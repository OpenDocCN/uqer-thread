**策略的原理是：**
1. 用 MACD 和 CCI 两个指标对50ETF(或者50指数，这个我没有试过)进行择时，即用这两个指标最常见的买卖信号；
2. 两个指标都看多的时候，卖出深度虚值的看跌期权；都看空的时候，卖出深度虚值的看涨期权；两个指标意见不和的时候，看震荡，双向卖出两份虚值期权。

因为期权优矿还没有回测框架，所以还没想好这样做回测的收益率该怎么算。

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from CAL.PyCAL import *
import seaborn as sns
sns.set_style("white")
import talib
```

```python
# 获取50ETF历史价格数据
cal = Calendar("China.SSE")
begin = Date(2015, 2, 9)
end = Date.todaysDate()
beginDate = cal.advanceDate(begin, "-40B")
beginDate = beginDate.toISO().replace("-", "")
endDate = end.toISO().replace("-", "")
etf_field = ["tradeDate", "openPrice", "closePrice", "highestPrice", "lowestPrice", "preClosePrice"]
etf_data = DataAPI.MktFunddGet(secID="510050.XSHG", beginDate=beginDate, endDate=endDate, field=etf_field, pandas="1")
etf_data.tail()
```

```python
# 使用talib计算50ETF的历史MACD和CCI指标值，参数使用默认参数
dif, dea, hist = talib.MACD(etf_data.closePrice.values, fastperiod=12, slowperiod=26, signalperiod=9)
cci = talib.CCI(etf_data.highestPrice.values, etf_data.lowestPrice.values, etf_data.closePrice.values)
etf_data["daily_return"] = etf_data["closePrice"]-etf_data["preClosePrice"]
etf_data["cci"] = cci
etf_data["macd_dif"] = dif
etf_data["macd_dea"] = dea
etf_data = etf_data[etf_data.tradeDate >= begin.toISO()]
etf_data["tradeDate"] = pd.to_datetime(etf_data["tradeDate"])
etf_data = etf_data.set_index("tradeDate")
etf_data = etf_data[["closePrice","macd_dif", "macd_dea", "cci", "daily_return"]]
```

```python
etf_data.tail()
```

```python
# 计算策略信号点MACD信号为dif>dea->做多；dif<dea->做空；CCI信号cci>100->做多;cci<-100->做空;-100<cci<100->盘整
etf_data["macd_sig"] = np.where((etf_data["macd_dif"]>=etf_data["macd_dea"]), 1, 0)
etf_data["macd_sig"] = np.where((etf_data["macd_dif"]<etf_data["macd_dea"]), -1, etf_data["macd_sig"])
etf_data["cci_sig"] = np.where(etf_data["cci"]>=100, 1, 0)
etf_data["cci_sig"] = np.where(etf_data["cci"]<-100, -1, etf_data["cci_sig"])
etf_data["etf_daily_pl"] = np.zeros_like(etf_data["cci"])
etf_data.head()
```

```python
# 只对两者同时的做多信号做多， 发出信号之后收益使用次日起的。
for i in range(len(etf_data.index)-1):
    if etf_data.ix[etf_data.index[i], "macd_sig"] == 1 and etf_data.ix[etf_data.index[i], "cci_sig"] == 1:
        etf_data.ix[etf_data.index[i+1], "etf_daily_pl"] = etf_data.ix[etf_data.index[i+1], "daily_return"]
```

```python
etf_data.tail()
```

```python
# 计算累计收益，同时乘以10000，计算绝对收益
etf_data["cumReturn"] = np.cumsum(etf_data.etf_daily_pl)
etf_data["cumReturn"] = etf_data["cumReturn"]*10000
etf_data.tail()
```

```python
# 画图
fig = plt.figure(figsize=(10,6))
ax = fig.add_subplot(1,1,1)

line1 = ax.plot(etf_data.index, etf_data.cumReturn, "r-", label=u"Return")
ax2 = ax.twinx()
line2 = ax2.plot(etf_data.index, etf_data.closePrice, "g-", alpha=.5, label=u"50ETF")
ax.grid()
ax.set_xlabel(u"tradeDate")
ax.set_ylabel(u"return")
ax2.set_ylabel(u"price")
line = line1+line2
label = [l.get_label() for l in line]
ax.legend(line, label)
plt.title(u"Return of MACD-CCI strategy")
```

50etf期权

```python
# 先去把所有的期权交易数据拿到
beginDate = Date(2015, 2, 9)
endDate = Date.todaysDate()
begin = beginDate.toISO().replace("-", "")
end = endDate.toISO().replace("-", "")
option_field = ["tradeDate", "exchangeCD","secID", "preClosePrice", "closePrice"]
opt_price_data = DataAPI.MktOptdGet(beginDate=begin, endDate=end, field = option_field, pandas="1")
```

```python
# 筛选出所有的50ETF期权价格数据
# 计算每个期权合约每天的绝对盈亏
# 从secID中提取contractType、strikePrice以及expDate
opt_price_data = opt_price_data[opt_price_data.exchangeCD == "XSHG"]
opt_price_data["daily_return"] = opt_price_data["preClosePrice"]-opt_price_data["closePrice"]
#opt_price_data["tradeDate"] = pd.to_datetime(opt_price_data["tradeDate"])
opt_price_data = opt_price_data.set_index("tradeDate")
opt_price_data["contractType"] = [x[6] for x in opt_price_data["secID"]]
opt_price_data["strikePrice"] = [x[13:17] for x in opt_price_data["secID"]]
opt_price_data["expDate"] = [x[7:11] for x in opt_price_data["secID"]]
opt_price_data.tail()
```

```python
# 在指定交易日中找到交易日当天最虚值的看涨和看跌期权合约（期限最长、虚值额最大）
def optDailyReturn(data, tradeDate, optionType):
    date = tradeDate
    # call
    if optionType == 1:
        # select cal
        select_cal = data[data.index == date]
        select_cal = select_cal[select_cal.contractType == "C"]
        select_cal = select_cal[select_cal.expDate == max(select_cal.expDate.tolist())]
        select_cal = select_cal[select_cal.strikePrice == max(select_cal.strikePrice.tolist())].daily_return.values[0]
        returns = select_cal
    # put
    if optionType == -1:
        # select put
        select_put = data[data.index == date]
        select_put = select_put[select_put.contractType == "P"]
        select_put = select_put[select_put.expDate == max(select_put.expDate.tolist())]
        select_put = select_put[select_put.strikePrice == min(select_put.strikePrice.tolist())].daily_return.values[0]
        returns = select_put
    return returns
```

```python
option_data = etf_data[["macd_sig", "cci_sig"]]
option_data["call_daily_return"] = np.zeros_like(option_data["macd_sig"])
option_data["put_daily_return"] = np.zeros_like(option_data["macd_sig"])
option_data["sig_return"] = np.zeros_like(option_data["macd_sig"])
option_data.tail()
```

```python
# 历遍每个交易日，找出每个交易日最虚值的看涨和看跌期权当日的绝对盈亏，保存下来
for i in range(len(option_data.index)):
    date = option_data.index[i].strftime("%Y-%m-%d")
    cal = optDailyReturn(opt_price_data, date, 1)
    option_data.ix[option_data.index[i], "call_daily_return"] = cal
    put = optDailyReturn(opt_price_data, date, -1)
    option_data.ix[option_data.index[i], "put_daily_return"] = put
```

```python
option_data.tail()
```

```python
for i in range(len(option_data.index)-1):
    # 多头信号，卖出2份最虚值看跌期权（期限最长，行权价最高）
    # 空头信号，卖出2份最虚值的看涨期权（期限最长，行权价最低）
    # 盘整信号， 卖出一份最虚值的宽跨式合约组合（一张最虚值的看涨和一张最虚值的看跌）
    # 因为都是根据收盘价信号，第二天做交易，因此暂时取信号第二天的合约收益率为我们账户收益率
    if option_data.ix[option_data.index[i], "macd_sig"] == 1 and option_data.ix[option_data.index[i], "cci_sig"] == 1:
        # 获取下一个交易日我们卖出的头寸的收益率当做我们的账户的下一个交易日的收益率
        option_data.ix[option_data.index[i+1], "sig_return"] = option_data.ix[option_data.index[i+1], "put_daily_return"] * 2
    elif option_data.ix[option_data.index[i], "macd_sig"] == 0 and option_data.ix[option_data.index[i], "cci_sig"] == 0:
        option_data.ix[option_data.index[i+1], "sig_return"] = option_data.ix[option_data.index[i+1], "call_daily_return"] * 2
    else:
        option_data.ix[option_data.index[i+1], "sig_return"] = option_data.ix[option_data.index[i+1], "call_daily_return"]+option_data.ix[option_data.index[i+1], "put_daily_return"]
```

```python
option_data["final_return"] = np.cumsum(option_data.sig_return)
option_data["final_return"] = option_data["final_return"]*10000
```

```python
option_data.tail()
```

```python
fig = plt.figure(figsize=(14,8))
ax = fig.add_subplot(1, 1, 1)
line1 = ax.plot(option_data.index, option_data.final_return, "r-", label=u"Option Return")
line3 = ax.plot(etf_data.index, etf_data.cumReturn, "b-", label=u"50ETF MACD-CCI Strategy")
ax2 = ax.twinx()
line2 = ax2.plot(etf_data.index, etf_data.closePrice, "g-", label=u"50ETF Price", alpha=.5)

line = line1+line2+line3
label = [l.get_label() for l in line]
ax.set_xlabel(u"tradeDate",fontsize=12)
ax.set_ylabel(u"profit",fontsize=12)
ax2.set_ylabel(u"price",fontsize=12)
plt.legend(line, label, loc="best")
ax.grid()
plt.title(u"50ETF Option Timing Seller Strategy  & 50ETF Timing Buyer Strategy", fontsize=16)
```
