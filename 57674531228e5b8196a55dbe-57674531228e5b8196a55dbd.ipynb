> * ##### 盈利因子 
> 	* ###### [ 'NetProfitRatio', 'OperatingProfitRatio', 'GrossIncomeRatio', 'TaxRatio', 'ROA', 'ROA5', 'ROE', 'ROE5', 'EGRO', 'EPS', 'DilutedEPS' ]
> 	* ##### 因子描述
> 		* ###### 'NetProfitRatio'：销售净利率，$NetProfitRatio = \frac{净利润}{营业收入}$
> 		* ###### 'OperatingProfitRatio'：营业利润率，$OperatingProfitRatio = \frac{营业利润}{营业收入}$
> 		* ###### 'GrossIncomeRatio'：销售毛利率，$GrossIncomeRatio = \frac{营业收入 - 营业成本}{营业收入}$
> 		* ###### 'TaxRatio'：销售税金率，$TaxRatio = \frac{营业税金及附加}{营业收入}$
> 		* ###### 'ROA'：资产回报率，$ROA = \frac{净利润}{总资产}$
> 		* ###### 'ROA5'：5年资产回报率，$ROA5 = \frac{过去5年平均净利润}{过去5年平均总资产}$
> 		* ###### 'ROE'：权益回报率，$ROE = \frac{净利润}{所有者权益}$
> 		* ###### 'ROE5'：5年权益回报率，$ROE5 = \frac{过去5年平均净利润}{过去五年平均所有者权益}$
> 		* ###### 'EGRO'：5年收益增长率，$EGRO = \frac{过去5年净利润关于时间的线性回归系数}{过去5年净利润均值的绝对值}$
> 		* ###### 'EPS'：每股收益，取自财报
> 		* ###### 'DilutedEPS'：稀释每股收益，取自财报
