# 利率期限结构与Nelson-Siegel模型

## 一、背景介绍

    本文主要对利率期限结构和常见的利率期限结构模型进行了介绍。利率期限结构，是指某个时间点上，不同期限的即期利率（或到期收益率）和到期期限之间的关系。具有相同风险、流动性及税收待遇，但期限不同的金融工具具有不同的利率水平。收益率曲线具有三个特征，研究者针对这三个特征提出了利率期限结构的三种理论。利率期限结构在债券的收益归因和风险归因里有着很多应用。本文的结构是：首先介绍利率期限结构的定义和其三个主要特征，然后介绍利率期限结构的三种理论，再对常见的利率期限结构模型进行讲解，最后着重介绍Nelson-Siegel模型。由于利率期限结构的研究体系庞大且深入，限于时间篇幅限制，在此作入门介绍，希望能起到抛砖引玉的作用。

## 二、利率期限结构理论

### 1. 关键概念
  - 到期收益率（Yield to Maturity）：投资者持有某一债券至到期，并假定其本金与利息都按时支付时，以买入价格计算的内部收益率（IRR）。
  - 即期利率（Spot Rate）：又称零息利率，是零息债券到期收益率的简称。
  - 利率期限结构（Term Structure of Interest Rate）：指某个时间点上，不同期限的即期利率（或到期收益率）和到期期限之间的关系。

### 2. 利率期限结构的形态

  - 常见的利率期限结构形态有水平、向上/向下倾斜和拱形：
    - ![![图片注释](http://storage-uqer.datayes.com/5a0d023c8bfebd0112e4088a/4159bcb6-ea0a-11e7-82c3-0242ac140002)](http://storage-uqer.datayes.com/5a0d023c8bfebd0112e4088a/4159bcb6-ea0a-11e7-82c3-0242ac140002)
    - 不同的利率期限结构形态反映了投资者对未来经济形势的预期。向下倾斜的收益率曲线表明投资者对未来经济形势不看好，投资者宁愿选择期限更长、收益率更低的债券，也不愿意购买短期限收益率高的债券，因为短期限债券在到期后，因经济形势走弱，面临比较大的再投资风险。

### 3. 利率期限结构理论

  - 一般来说，收益率曲线有三个特征：
    - 不同期限的债券收益率有同向运动的趋势
    - 收益率曲线通常倾向于向上倾斜
    - 短期债券收益率的波动通常要比长期债券收益率的波动大

  - 为了解释这些特征，有三个主要的利率期限结构理论：
    - 预期理论：
      - 预期理论假设投资者对于债券的期限没有偏好，债券之间可以完全替代，即两个连续的1年期债券组合和一个2年期的债券，若两者的期望收益相等，则投资者对这两者没有偏好。预期理论在假设中剔除了利率风险和再投资风险。
      - 长期利率是短期利率期望值的平均：
        $i_nt = {{i_t + i_{t+1}^e + ... + i_{t+n-1}^e} \over n}$
      - 由上述公式可以看出， $i_t$的变动会引起$i_{nt} (n = 2, 3, 4...)$的变动，解释了收益率曲线的特征1）。求平均对于波动有平滑作用，因此期限越长的利率波动越小。但是预期理论难以解释为什么收益率曲线通常向上倾斜。

    - 分割市场理论：
      - 分割市场理论假设不同期限的债券所在的市场之间完全无关。对于一支债券来说，其它不同期限的债券的预期收益高低对其没有影响，该债券的价格仅由供需决定。
      - 分割市场理论考虑了债券面临的利率风险和再投资风险。债券的期限越长，面临的利率风险和再投资风险也越高，因此长期债券的收益率要高于短期债券，可以解释为什么收益率曲线通常向上倾斜。
      - 但是由于有着不同期限债券市场无关的假设，分割市场理论难以解释收益率的同向运动和短期利率波动大于长期利率波动的情况。
  
    - 流动性偏好假说：
      - 流动性偏好假说结合了前面两种理论的优点，即假设不同债券之间存在一定的可替代性，并且考虑利率风险和再投资风险：
        $i_nt = {{i_t + i_{t+1}^e + ... + i_{t+n-1}^e} \over n} + l_{nt}$
      - 对于期限越长的债券，其面临的利率风险和再投资风险也就越高。因此投资者若要持有更长期限的债券，就会要求更高的风险溢价。由于结合了前面两种理论，流动性偏好假说能够较好的解释收益率曲线的三个特点。

## 三、利率期限结构模型

### 1. 利率期限结构模型分类

  - ![图片注释](http://storage-uqer.datayes.com/5a0d023c8bfebd0112e4088a/d0024406-ea0a-11e7-ba0d-0242ac140002)
  - 利率期限结构模型主要有动态模型和静态模型两类。静态模型着重于在有限数据样本下，找到对数据最佳拟合的收益率曲线。接下来要详细介绍的Nelson-Siegel模型就属于静态模型。动态模型又分为均衡模型和无套利模型两类。均衡模型一般从经济理论出发，根据市场的均衡条件得出利率须遵循的过程。代表模型有Vasicek模型和Cox-Ingersoll-Ross模型；无套利模型则是用市场上的资产价格，根据各个资产之间必须满足的无套利条件来推导利率的发展过程。代表模型有Hull-White模型等。

### 2. 均衡模型

  - Vasicek模型
    - Vasicek模型假设短期利率遵循如下随机偏微分方程： $ dr_t = \kappa(\mu - r_t)dt + \sigma dW_t $。可以看到Vasicek模型假设短期利率的变动服从均值回归，$\kappa$代表利率回归均值的调整速度，$\mu$代表利率的长期水平，$\sigma$代表利率的波动率， $dW_t$为布朗运动。在时间点t，若利率水平大于$\mu$，则$\kappa(\mu - r_t)$为负值，反之亦然。
    - Vasicek模型的偏微分方程中，利率的波动率$\sigma$为常数，因此当利率水平接近0的时候，$dr_t$若为负值，会造成下一时间点的利率为负值。通常我们都假设利率非负，因此Vasicek模型存在利率有可能为负值的问题。

  - Cox-Ingersoll-Ross模型
    - CIR模型假设短期利率遵循如下随机偏微分方程： $dr_t = \kappa(\mu - r_t)dt + \sigma \sqrt{r_t}dW_t$
    - CIR模型在Vasicek模型的基础上进行了改进，$\sqrt{r_t}$的加入，使得$\sigma \sqrt{r_t}dW_t$这一项随着$r_t$接近0而同样趋近于0，这在一定程度上避免了负利率的出现。

  - 对于均衡模型来说，一旦定义了利率的发展过程，也就完全定义了初始期限结构和该期限结构在未来时刻的演变。均衡模型的形式较为简单，模型参数是与时间无关的常数，难以准确刻画利率变化的客观规律。在均衡模型里，初始利率期限结构是模型的输出而不是输入，因此会与初始观察到的市场利率期限结构有冲突。

### 3. 无套利模型

  - Hull-White模型
    - Hull-White模型假设短期利率遵循如下随机偏微分方程： $dr(t) = \big[\theta(t) - \alpha(t)r(t)\big]dt + \sigma(t)dW(t)$
    - 在Hull-White模型中，$\theta$、$\alpha$和$\sigma$都是时间的函数。Hull-White的贡献是在Vasicek和CIR的基础上，引入了和时间有关的参数，能更准确的描述利率变化的规律。

## 四、Nelson-Siegel模型

### 1. 模型简介 

  - 由Charles Nelson和Andrew Siegel在1987年提出，用参数模型来拟合现有的收益率曲线：
  - $y(\tau) = \beta_1 + \beta_2\bigg[{{1 - exp(-\tau/\lambda)}\over{\tau/\lambda}}\bigg] + \beta_3\bigg[{{1 - exp(-\tau/\lambda)}\over{\tau/\lambda}} - exp(-\tau/\lambda)\bigg]$
  - $y(\tau)$为即期利率，$\tau$为债券的期限，$\beta_1、\beta_2、\beta_3$和$\lambda$为模型的四个待估计参数。

### 2. 参数含义

  - $y(\tau) = \beta_1 + \beta_2\bigg[{{1 - exp(-\tau/\lambda)}\over{\tau/\lambda}}\bigg] + \beta_3\bigg[{{1 - exp(-\tau/\lambda)}\over{\tau/\lambda}} - exp(-\tau/\lambda)\bigg]$
  - 当$\tau$趋近于$\infty$时，$\lim_{\tau \to \infty}y(\tau) = \beta_1$。因此$\beta_1$代表了利率的长期水平，$\beta_1$的变动会使整条曲线的高度上下变动，因而被称为水平参数。
  - 当$\tau$趋近于0时，$\lim_{\tau \to 0}y(\tau) = \beta_1 + \beta_2 \Longrightarrow \lim_{\tau \to \infty}y(\tau) - \lim_{\tau \to 0}y(\tau) = -\beta_2 $。$\beta_2$被称为斜率参数，主要对短期利率产生影响。
  - 当$\tau$趋于正无穷或趋于0时，$\bigg[{{1 - exp(-\tau/\lambda)}\over{\tau/\lambda}} - exp(-\tau/\lambda)\bigg]$均为0（通过洛必达法则可解），因此$\beta_3$的变动对短期和长期利率的影响较小，主要对中期利率产生影响。$\beta_3$决定着曲线的曲率，被称为曲率参数。
  - $\lambda$控制者斜率参数和曲率参数的指数增长率。$\lambda$越小，指数项的衰减越慢，对于长期限的数据点拟合越好；$\lambda$越大，指数项的衰减越快，对于短期限的数据点拟合越好。

### 3. 参数求解

  - 让$\theta = (\beta_1, \beta_2, \beta_3, \lambda)$，对于债券i的价格$P_i$有：
    $\overline{P_i} (\theta) = \sum_{j=1}^n C_{ij}d(t_{ij}, \theta) =  \sum_{j=1}^n C_{ij}e^{-t_{ij}Z(t_{ij}, \theta)}$
    其中，$C_{ij}$为债券i的第j次现金流，$e^{-t_{ij}Z(t_{ij}, \theta)}$为第j次现金流的折现率，$Z(t_{ij}, \theta)$是通过Nelson-Siegel模型计算出来的即期利率，也就是上文中的$y(\tau)$。
  - 我们要最小化的目标函数为:  
    $Obj(P) = \sum_{i=1}^n w_i(P_i - \overline{P_i})^2$  
    $s.t. \beta_1 > 0, \beta_1 + \beta_2 > 0$
  - $w_i$是债券i的权重，若要保证中短期限债券价格的拟合优度，可将$w_i$设为： $w_i = {1/D_i \over \sum_{i=1}^n 1/D_i}$，$D_i$为债券i的久期。$P_i$为市场上观察到的债券i的价格，通过最小二乘法可求得Nelson-Siegel的参数。

### 4. 基准利率曲线变动对债券价格的影响：

 - 债券的收益一般可以拆解为息票收益和资本利得两部分。其中资本利得可以进一步拆解为由基准利率曲线变动带来的收益和利差变动带来的收益。
 - 通过Nelson-Siegel模型，我们可以更进一步，将基准利率曲线的变动继续拆解为平移、扭动和蝶形。对于$t_0$和$t_1$两个时间点：
 - $\Delta y_{平移} = \beta_1^1 - \beta_1^0$  
 - 平移变动指利率曲线的整体上下移动，向上平移会减少各久期债券的资本利得收益，向下平移会增加各久期债券的资本利得收益。
 - $\Delta y_{扭动} = \beta_2^1{{1 - e^{-t_0 \over \tau}} \over {t_0 \over \tau}} - \beta_2^0{{1 - e^{-t_0 \over \tau}} \over {t_0 \over \tau}}$
 - 扭动变化是指利率期限结构斜率的变动，长短两端的利率如果发生了相反方向的变动，或者一端的变动幅度大于另一端，都会引起利率曲线的扭动。
 - $\Delta y_{蝶形} = \beta_3^1\bigg[{{{1 - e^{-t_0 \over \tau}}\over {t_0 \over \tau}} - e^{-t_0 \over \tau}}\bigg]- \beta_3^0\bigg[{{{1 - e^{-t_0 \over \tau}}\over {t_0 \over \tau}} - e^{-t_0 \over \tau}}\bigg]$
 - 蝶形变化指利率曲线弯曲程度发生了变化，如果曲线更为凸向原点，长短端资产面临的利率环境不如中间期限资产，反之亦然。

### 5. 市场利率变动和修正久期的关系

  - 对于修正久期，我们经常能看到这么一句话：“假设债券的修正久期为MD,那么当利率变动100bp，债券的价格就会反向变动MD%”。乍看之下这句话好像没问题，但是请仔细想一想，我们在计算修正久期的时候，里面用到的只有该债券的到期收益率（YTM）。而这句话中的“利率”一般指的是即期利率，这岂不是概念混淆了吗？
  - 其实上面这句话是在一个假设之下才成立的，而人们往往忽略掉了这个假设：各个期限的即期利率都变动同样方向同样幅度。这个假设也是比较符合收益率曲线的特征的。
  - 到期收益率可以看作是即期利率的加权平均，在这里我们不需要知道权重是多少：$YTM = \sum_{i=1}^nw_iZ(t_i)$
  - 按照上面的假设，所有期限的即期利率都变动$\Delta$，我们有： $YTM' = \sum_{i=1}^nw_i(Z(t_i) + \Delta) = \sum_{i=1}^nw_iZ(t_i) + \Delta\sum_{i=1}^nw_i = YTM + \Delta$
  - 我们可以看到，此时若即期利率变动$\Delta$，债券的到期收益率也会变动$\Delta$。因此我们可以说：“假设债券的修正久期为MD,那么当利率变动100bp，债券的价格就会反向变动MD%”。
  - 上一节中我们将基准利率的变动拆解为了平移、扭动和蝶形三种，利用修正久期的定义，若即期利率变动$\Delta y$：
  - ${\Delta p \over P} = -MD\Delta y + {1\over2}C(\Delta y)^2 = -MD(\Delta y_{平移} + \Delta y_{扭动} + \Delta y_{蝶形}) + {1\over2}C(\Delta y_{平移} + \Delta y_{扭动} + \Delta y_{蝶形})^2$

### 6. Nelson-Siegel-Svensson模型

  - Nelson-Siegel模型虽然被多国央行采用，但是它还是有一些缺点，比如只有一个曲率参数，因此对于具有多个局部最大和最小的曲线拟合效果不好。
  - Svensson在Nelson-Siegel的三因子模型基础上，加入了第四个因子：
  - $y(\tau) = \beta_1 + \beta_2\bigg[{{1 - exp(-\tau/\lambda)}\over{\tau/\lambda}}\bigg] + \beta_3\bigg[{{1 - exp(-\tau/\lambda_1)}\over{\tau/\lambda_1}} - exp(-\tau/\lambda_1)\bigg] + \beta_4\bigg[{{1 - exp(-\tau/\lambda_2)}\over{\tau/\lambda_2}} - exp(-\tau/\lambda_2)\bigg]$
  - 其实就是加了第二个曲率参数，能更好的拟合一些曲线。
  - 但是NSS模型也存在一个大问题，也许你已经看出来了。当$\lambda_1$和$\lambda_2$取值接近的时候，NSS模型就存在很强的多重共线性。此时若进行线性回归，我们最多能得到$\lambda_1 + \lambda_2$的值，但是很难将两者分开。
  - 因此有人提出了NSS的调整模型：
  - $y(\tau) = \beta_1 + \beta_2\bigg[{{1 - exp(-\tau/\lambda)}\over{\tau/\lambda}}\bigg] + \beta_3\bigg[{{1 - exp(-\tau/\lambda_1)}\over{\tau/\lambda_1}} - exp(-\tau/\lambda_1)\bigg] + \beta_4\bigg[{{1 - exp(-\tau/\lambda_2)}\over{\tau/\lambda_2}} - exp(-{\color{red}2}\tau/\lambda_2)\bigg]$
  - 其实就是在最后一项的$\tau$前面加了个2，这样让第三项和第四项不同，避免了共线性问题。加了2之后，第四项的指数衰减速度要比第三项快。

## 五、 总结
  - 模型优点：
	  - 简约模型，对于常见收益率曲线描述性较强
      - 参数有较好的经济学解释性
  - 模型缺点：
	  - 对于一些特定的收益率曲线无法准确描述，如存在多个局部最大或局部最小的曲线
      - 参数之间存在相关性，因此模型会有一定的共线性。如Nelson-Siegel-Svensson模型中，若两个$\lambda$参数取相同的值，后两个参数就完全线性相关
