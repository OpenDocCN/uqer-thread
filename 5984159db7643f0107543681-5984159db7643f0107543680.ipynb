# 2-PBPE因子分位绘图

```python
# 本代码目的：获取PB，PE，PS等因子,按时间存入矩阵，并绘制分位数图，直观了解市场或板块的当前估值状态

import numpy as np
import pandas as pd
import datetime
import matplotlib.pyplot as plt
from CAL.PyCAL import *


# part1. 基本信息赋值
cal = Calendar('China.SSE')
begin = datetime.date(2007,1,1)                  # 设定起始时间
end = datetime.date(2017,8,1)                    # 设定终止时间
#universe = DynamicUniverse('HS300')             # 设定股票池，动态沪深300，无未来信息。如果用于策略回测，要用这个函数
#universe = set_universe('HS300',"2012-01-01")   # 设定股票池,指定起始时点的标的范围，无未来信息问题
universe = set_universe('HS300')                 # 设定股票池，有未来信息问题，但不用于量化策略回测，所以是可以的。创业板-'CYB'；中小板-'ZXB'；上证50-'SH50'；A股-'A'

Tdays = cal.bizDatesList(begin, end)             # 得到起止时间之间的交易日列表

#Qt = [0.5', '0.7','0.8','0.9','0.95']   #可以设置分位数数量,从最低的5%分位到最高的95%分位
Qt = ['0.8','0.9','0.95'] 

#每天PB分位数所对应的矩阵初始化
Quantiles = pd.DataFrame(index = Qt, columns = Tdays)


# part2. 计算分位数
for d in Tdays:                          # 遍历Tdays时间列表中的日期
    Today = d
    #可以修改PB为其他因子，如PE、PS等等
    df = DataAPI.MktStockFactorsOneDayGet(tradeDate=d,secID=universe,field=u"tradeDate,PB").set_index("tradeDate")
    # 获取股票池中某一天所有股票的PB值（若要更换因子，在此处将PB换为其他因子），并把日期设为索引
    #factor_quantiles = df.dropna().quantile([0.5, 0.7, 0.8, 0.9, 0.95])
    factor_quantiles = df.dropna().quantile([0.8, 0.9, 0.95])
    # 去掉缺省值，计算得到某一天的分位数，保存至factor_quantiles
    if Today in Quantiles.columns:      # 遍历时间列表
        Quantiles[Today] = factor_quantiles.values            # 将某一天的分位数列表转存至统一的列表中

# part3. 绘图
plt.figure(figsize=(24,6))              # 设定图的大小
# 绘制相应的5条分位数曲线
# Quantiles.columns对应的是所有交易日，Quantiles.values[0](1,2,3,4）分别对应不同的分位数
p1 = plt.plot(range(len(Quantiles.columns)),Quantiles.values[0])
p2 = plt.plot(range(len(Quantiles.columns)),Quantiles.values[1])
p3 = plt.plot(range(len(Quantiles.columns)),Quantiles.values[2])
#p4 = plt.plot(range(len(Quantiles.columns)),Quantiles.values[3])
#p5 = plt.plot(range(len(Quantiles.columns)),Quantiles.values[4])
    
plt.title("PB Distribution Over Time", fontsize=16)          # 设定图的名字，并设定字体大小
#plt.legend(['0.5_Q','0.7_Q','0.8_Q','0.9_Q','0.95_Q'])      # 给出图标的图例名称
plt.legend(['0.8_Q','0.9_Q','0.95_Q'])                       # 给出图标的图例名称
x_tick = []                                                  # 给出图例中横轴图例

for i in range(len(Quantiles.columns)):                      # 遍历日期数目
    if i%30 == 0:                                            # 每三十个交易日期生成一个图例
        x_tick.append(Quantiles.columns[i])
    else: x_tick.append("")                                  # 其余图例为空值

plt.xticks(range(len(x_tick)), x_tick,rotation = -90)
plt.xlim(0,len(x_tick))                                      # 展示时间范围
plt.show()                                                   # 出图

#打印当前分位数所对应的PB、PE、PS值
print Quantiles.icol(-1)

```
