```python
# 判断是否符合条件
def check_MACD_chgPct(x):
    if x.MACD0<=0 and x.MACD1>0 and x.chgPct>0:
        return(1)
    elif x.MACD0<=0 and x.MACD1>0 and x.chgPct<0:
        return(0)
    elif x.chgPct>0:
        return(-1)
    else:
        return(-2)
    
    
```

```python

import pandas as pd
import numpy as np
import math

codeList = ['000002', '000009', '000069', '000301', '000502', '000629', '000630', '000729', '000898', '000930']
start = '20200101'
end = '20201201'
# start = '20211001'
# end = '20211015'
tickerList = ','.join(codeList)


# T日收盘确定指标 T+1收盘确定指标变化 T+2日确定涨跌幅
# 制作日期列表、
df = DataAPI.TradeCalGet(exchangeCD=u"XSHG",beginDate=start,endDate=end,isOpen=u"1",field=u"",pandas="1")
# print(df)

dateList = df.loc[:,'calendarDate'].values.tolist()
dateList = [x[:4]+x[5:7]+x[8:] for x in dateList]
print(dateList)
# T日收盘确定指标 T+1收盘确定指标变化 T+2日确定涨跌幅
dateList0 = dateList[:-2]
dateList1 = dateList[1:-1]
dateList2 = dateList[2:]

dateList0 = dateList[:-3]
dateList1 = dateList[1:-2]
dateList2 = dateList[3:]

dateList = [dateList0,dateList1,dateList2]
print(dateList)

# for i in range(len(dateList[0])):
#     print(dateList[0][i],dateList[1][i],dateList[2][i])
#     print('*')

# 样本列表
dfList = []
# 逐日收集样本
for i in range(len(dateList[0])):
    df0 = DataAPI.MktStockFactorsOneDayGet(ticker=tickerList,tradeDate=dateList[0][i],field=u"MACD,ticker",pandas="1")
    # print(df0)
    df1 = DataAPI.MktStockFactorsOneDayGet(ticker=tickerList,tradeDate=dateList[1][i],field=u"MACD,ticker",pandas="1")
    # print(df1)
    df2 = DataAPI.MktEqudGet(ticker=tickerList,tradeDate=dateList[2][i],field=u"chgPct,ticker",pandas="1")
    # print(df2)
    
    df0 = df0.rename(columns={'MACD':'MACD0'})
    df1 = df1.rename(columns={'MACD':'MACD1'})
    df = df0.merge(df1,on='ticker',how='left')
    df = df.merge(df2,on='ticker',how='left')
    df = df.loc[:,['MACD0','MACD1','chgPct']]
    # print(df)
    
    # 标记样本
    df['TF'] = df.apply(lambda x: check_MACD_chgPct(x),axis=1) 
    # print(df)
    dfList.append(df)

allDf = pd.concat(dfList)
print(allDf)

# 区分两类样本
T0DF = allDf[allDf.TF>=0]
T1DF = allDf[allDf.TF<0]

print(T0DF.shape)
print(T1DF.shape)
DATA0 = T0DF.loc[:,'chgPct'].values.tolist()
DATA1 = T1DF.loc[:,'chgPct'].values.tolist()

# 做双总体均值的T检验
```

```python
# 双总体T检验
import scipy

# 按照T检验的步骤，首先做方差齐次性检验，下方代码就是在做方差齐次性检验，一行代码，so easy。
# print(DATA0)

A = scipy.stats.levene(DATA0,DATA1)
# A = scipy.stats.bartlett(DATA0,DATA1)
print(A)
# 得到的结果如下所示： LeveneResult(statistic=1.9946269989379286, pvalue=0.15800013610776201) 这里我们只需要关注pvalue，P=0.15大于0.05，证明方差齐次性检验的原假设成立，说明两组样本方差满足齐次性。

# 双样本T检验
# 接着做双样本T检验，依然是一行代码。注意,当两组样本之间的方差不满足齐次性，需要加一个参数equal_var=False。此API默认为equal_var=True。

if A[0]>0.05:
    B = scipy.stats.ttest_ind(DATA0,DATA1,equal_var=True)
    print('双样本方差齐次')
else:
    B = scipy.stats.ttest_ind(DATA0,DATA1,equal_var=False)
    print('双样本方差不齐次')
print(B)

# P=0.2780大于0.05，证明双样本T检验的原假设成立，即两个总体之间的均值不存在差异
if B[0]>0.05:
    print('双样本均值不存在显著差异')
else:
    print('双样本均值存在显著差异')

print('mean_DATA0:')
print(sum(DATA0)/len(DATA0))
print('std_DATA0:')
print(np.std(DATA0))
print('mean_DATA1:')
print(sum(DATA1)/len(DATA1))
print('std_DATA0:')
print(np.std(DATA1))
```
