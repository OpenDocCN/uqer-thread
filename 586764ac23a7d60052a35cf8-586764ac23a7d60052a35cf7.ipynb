# 因子处理

- 这里使用之前处理过的文档，要是没看过之前的帖子，传送门见文档开头

```python
TVSTD20_deal = pd.read_csv('TVSTD20_deal.csv')
TVSTD20_deal['tradeDate'] = map(Date.toDateTime, map(DateTime.parseISO, TVSTD20_deal['tradeDate']))
TVSTD20_deal.set_index('tradeDate',inplace=True)

ATR6_deal = pd.read_csv('ATR6_deal.csv')
ATR6_deal['tradeDate'] = map(Date.toDateTime, map(DateTime.parseISO, ATR6_deal['tradeDate']))
ATR6_deal.set_index('tradeDate',inplace=True)

MoneyFlow20_deal = pd.read_csv('MoneyFlow20_deal.csv')
MoneyFlow20_deal['tradeDate'] = map(Date.toDateTime, map(DateTime.parseISO, MoneyFlow20_deal['tradeDate']))
MoneyFlow20_deal.set_index('tradeDate',inplace=True)

Boll_deal = pd.read_csv('Boll_deal.csv')
Boll_deal['tradeDate'] = map(Date.toDateTime, map(DateTime.parseISO, Boll_deal['tradeDate']))
Boll_deal.set_index('tradeDate',inplace=True)
```

- 注意，这里必须理清“东部王国”和“北方战神”的微妙关系……咳咳，是Moneyflow和Tvstd……（**从量纲角度来看他们其实是同一个物种，只是观察的角度不同**），简单使用等权处理（可以用更加统计学的方式，如协方差，机器学习，不过那些我自己还没搞清楚……就不班门弄斧了）

```python
muti_factor_X = pd.DataFrame(index=TVSTD20_deal.index,columns=TVSTD20_deal.columns)
for tradeDate in TVSTD20_deal.index:
    muti_factor_X.ix[tradeDate] = (TVSTD20_deal.ix[tradeDate] + ATR6_deal.ix[tradeDate]/2 + MoneyFlow20_deal.ix[tradeDate] + Boll_deal.ix[tradeDate]/2)/4.0
muti_factor_X.to_csv('muti_factor_X.csv')
```

```python
#读取数据并处理
muti_factor_X = pd.read_csv('muti_factor_X.csv') 
muti_factor_X['tradeDate'] = map(Date.toDateTime, map(DateTime.parseISO, muti_factor_X['tradeDate']))
muti_factor_X.set_index('tradeDate',inplace=True)
```

```python
# 因子历史表现
n_quantile = 10
# 统计十分位数
cols_mean = ['meanQ'+str(i+1) for i in range(n_quantile)]
cols = cols_mean
corr_means = pd.DataFrame(index=muti_factor_X.index, columns=cols)

# 计算相关系数分组平均值
for dt in corr_means.index:
    qt_mean_results = []

    # 相关系数去掉nan和绝对值大于1的
    tmp_factor = muti_factor_X.ix[dt].dropna()
    #tmp_factor = tmp_factor[(tmp_factor<=1.0) & (tmp_factor>=-1.0)]
    
    pct_quantiles = 1.0/n_quantile
    for i in range(n_quantile):
        down = tmp_factor.quantile(pct_quantiles*i)
        up = tmp_factor.quantile(pct_quantiles*(i+1))
        mean_tmp = tmp_factor[(tmp_factor<=up) & (tmp_factor>=down)].mean()
        qt_mean_results.append(mean_tmp)
    corr_means.ix[dt] = qt_mean_results


# ------------- 因子历史表现作图 ------------------------

fig = plt.figure(figsize=(12, 6))
ax1 = fig.add_subplot(111)

lns1 = ax1.plot(corr_means.index, corr_means.meanQ1, label='Q1')
lns2 = ax1.plot(corr_means.index, corr_means.meanQ5, label='Q5')
lns3 = ax1.plot(corr_means.index, corr_means.meanQ10, label='Q10')

lns = lns1+lns2+lns3
labs = [l.get_label() for l in lns]
ax1.legend(lns, labs, bbox_to_anchor=[0.5, 0.1], loc='', ncol=3, mode="", borderaxespad=0., fontsize=12)
ax1.set_ylabel(u'因子', fontproperties=font, fontsize=16)
ax1.set_xlabel(u'日期', fontproperties=font, fontsize=16)
ax1.set_title(u"因子历史表现", fontproperties=font, fontsize=16)
ax1.grid()
```

- 因子的历史表现比较平稳，这点相比之前单因子有较好的提升，但5分位和10分位的距离太近还是对合成因子选股造成一定影响。

```python
forward_10d_return_data = pd.read_csv('ForwardReturns_10D.csv')
forward_10d_return_data['tradeDate'] = map(Date.toDateTime, map(DateTime.parseISO, forward_10d_return_data['tradeDate']))
forward_10d_return_data.set_index('tradeDate',inplace=True)
forward_10d_return_data = forward_10d_return_data[forward_10d_return_data.columns[1:]]
```

```python
ic_data = pd.DataFrame(index=muti_factor_X.index, columns=['IC','pValue'])#计算非流动性因子与收益率的秩相关系数

# 计算相关系数
for dt in ic_data.index:
    tmp_illiq = muti_factor_X.ix[dt]
    tmp_ret = forward_10d_return_data.ix[dt]
    cor = pd.DataFrame(tmp_illiq)
    ret = pd.DataFrame(tmp_ret)
    cor.columns = ['corr']
    ret.columns = ['ret']
    cor['ret'] = ret['ret']
    cor = cor[~np.isnan(cor['corr'])][~np.isnan(cor['ret'])]
    if len(cor) < 5:
        continue

    ic, p_value = st.spearmanr(cor['corr'],cor['ret'])   # 计算秩相关系数 RankIC
    ic_data['IC'][dt] = ic
    ic_data['pValue'][dt] = p_value

ic_data.to_csv('muti_factor_X_ic.csv')

print 'mean of IC: ', ic_data['IC'].mean()
print 'median of IC: ', ic_data['IC'].median()
print 'the number of IC(all, plus, minus): ', (len(ic_data), len(ic_data[ic_data.IC>0]), len(ic_data[ic_data.IC<0]))
```

```python
#分组超额收益分析
print '======================='
start_time = time.time()

n_quantile = 40
# 统计十分位数
cols_mean = [i+1 for i in range(n_quantile)]
cols = cols_mean

excess_returns_means = pd.DataFrame(index=muti_factor_X.index, columns=cols)

# 计算ILLIQ分组的超额收益平均值
for dt in excess_returns_means.index:
    qt_mean_results = []
    
    # ILLIQ去掉nan
    tmp_factor = muti_factor_X.ix[dt].dropna()
    tmp_return = forward_10d_return_data.ix[dt].dropna()
    tmp_return_mean = tmp_return.mean()
    
    pct_quantiles = 1.0/n_quantile
    for i in range(n_quantile):
        down = tmp_factor.quantile(pct_quantiles*i)
        up = tmp_factor.quantile(pct_quantiles*(i+1))
        i_quantile_index = tmp_factor[(tmp_factor<=up) & (tmp_factor>=down)].index
        mean_tmp = tmp_return[i_quantile_index].mean() - tmp_return_mean
        qt_mean_results.append(mean_tmp)
        
    excess_returns_means.ix[dt] = qt_mean_results

excess_returns_means.dropna(inplace=True)

finish_time = time.time()
print ''
print str(finish_time-start_time) + ' seconds elapsed in total.'
```

# 策略实现

- 感谢之前矿友**中华田园犬**的提示，这里加入去除新股（代码是优矿Knowledge里的，知识库内容丰富啊），有些pro的内容需要联系优矿秘书申请才能使用，妥妥的干货

```python
import datetime
from CAL.PyCAL import *
cal = Calendar('China.SSE')

def st_remove(source_universe, st_date=None):
    """
    给定股票列表,去除其中在某日被标为ST的股票
    Args:
        source_universe (list of str): 需要进行筛选的股票列表
        st_date (datetime): 进行筛选的日期,默认为调用当天
    Returns:
        list: 去掉ST股票之后的股票列表

    Examples:
        >> universe = set_universe('A')
        >> universe_without_st = st_remove(universe)
    """
    st_date = st_date if st_date is not None else datetime.datetime.now().strftime('%Y%m%d')
    df_ST = DataAPI.SecSTGet(secID=source_universe, beginDate=st_date, endDate=st_date, field=['secID'])
    return [s for s in source_universe if s not in list(df_ST['secID'])]

def new_remove(ticker,tradeDate= None,day = 1):
    """
    给定股票列表,去除某个日期前多少个交易日,之后上市的新股
    Args:
        ticker (list of str): 需要进行筛选的股票列表（无后缀）
        tradeDate (datetime): 进行筛选的日期,默认为调用当天
        day (int): 向前漂移的交易日的个数
    Returns:
        list: 去掉新股股票之后的股票列表（无后缀）

    Examples:
        >> universe = set_universe('A')
        >> universe = [stk[0:6] for stk in universe]
        >> universe_without_new = new_remove(universe)
    """
    tradeDate = tradeDate if tradeDate is not None else datetime.datetime.now()
    period = '-' + str(day) + 'B'
    pastDate = cal.advanceDate(tradeDate,period)
    pastDate = pastDate.strftime("%Y-%m-%d")
    ipo_date = DataAPI.SecIDGet(partyID=u"",assetClass=u"e",ticker=ticker,cnSpell=u"",field=u"ticker,listDate",pandas="1")
    remove_list = ipo_date[ipo_date['listDate'] > pastDate]['ticker'].tolist()
    return [stk for stk in ticker if stk not in remove_list]

def ticker2secID(ticker):
    """
    给定股票代码转化为股票内部编码
    Args:
        tickers (list): 需要转化的股票代码列表
    Returns:
        list: 转化为内部编码的股票编码列表

    Examples:
        >> tickers = ['600000','000001']
        >> secIDs = ticker2secID(tickers)
    """
    universe = DataAPI.EquGet(equTypeCD=u"A",listStatusCD="L,S,DE,UN",field=u"ticker,secID",pandas="1") # 获取所有的A股（包括已退市）
    universe = dict(universe.set_index('ticker')['secID'])
    if isinstance(ticker, list):
        res = []
        for i in ticker:
            if i in universe:
                res.append(universe[i])
            else:
                print i, ' 在universe中不存在，没有找到对应的secID！'
        return res
    else:
        raise ValueError('ticker should be list！')
```
