# pandas的索引

```python
import numpy as np 
import pandas as pd
df = pd.DataFrame(np.random.randn(6, 4), index = list('abcdef'), columns = list('ABCD'))
df
```

loc是对index的名称索引

```python
df.loc['a']
```

```python
df.loc[['a','c']]
```

```python
df.loc['a':'c']
```

```python
df['A']
```

```python
df['A'].loc['a']
```

iloc是按照index的位置来索引

```python
df.iloc[0]
```

```python
df.iloc[[0, 2]]
```

```python
df.iloc[0:3]
```

```python
df['A'].iloc[0]
```

ix是按照index的位置和名称混合索引，它会自动识别是位置还是名称，非常方便

```python
df.ix[0]
```

```python
df.ix['a']
```

```python
df.ix['a']['A']
```

```python
df.ix[df['A'] > 0]
```

```python
df.ix[(df['A'] > 0) & (df['B'] > 0)]
```

```python
df.ix[df.index == 'a']
```

```python
df['E'] = list('onetwo')
df
```

```python
df.ix[df['E'] == 'o']
```

获得某一条件下的index

```python
df.index[df['E'] == 'n']
```

```python
df.index[df['E'] == 'n'].values[0]
```
