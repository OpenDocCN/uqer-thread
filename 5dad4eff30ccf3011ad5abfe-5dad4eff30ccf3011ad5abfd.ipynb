````python
def get_historical_closeIndex(ticker, start, end):
    # get the data for the tickers.  This will be a panel
    p = DataAPI.MktIdxdGet(indexID=ticker,beginDate=start,endDate=end,field=u"tradeDate,indexID,closeIndex",pandas="1")
    p['tradeDate'] = pd.to_datetime(p['tradeDate']) #因为DataAPI所有日期都是以字符串形式存储的，因此此处需要转换成日期时间型,这里很重要！
    # rename the columns
    p.rename(columns={'tradeDate': 'Date', 
                      'closeIndex': 'Close'}, inplace=True)
    # pivot each ticker to a column
    pivoted = p.pivot(index='Date', columns='indexID')
    # and drop the one level on the columns
    pivoted.columns = pivoted.columns.droplevel(0)
    return pivoted
````

```
融资融券交易，是指投资者向具有融资融券业务资格的证券公司提供担保物，借入资金买入证券（融资交易）或借入证券并卖出（融券交易）的行为。包括券商对投资者的融资、融券和金融机构对券商的融资、融券。
```

`融资融券余额`

```
融资余额指融资买入股票额与偿还融资额的差额。融资余额若增加时，表示投资者心态偏向买方，市场人气旺盛，属强势市场；反之则属于弱势市场。融券余额指融券卖出额与每日偿还融券额的差额。融券余额增加，表示市场趋向卖方市场；反之则趋向买方。
```


````python
def rong_zi_rong_quan():
    """融资融券余额信息
    
    目前两融数据每个交易日9:30点左右才更新。
    """
    trade_days = report.trade_pre_n_days(n=2)
    data = DataAPI.FstTotalGet(exchangeCD=u"XSHE,XSHG", beginDate=trade_days[0])
    data = data.groupby(by='tradeDate').sum()
    data['tradeVal'] = data['tradeVal'] / (10**12) 
    ax = data.plot(y=['tradeVal'], kind='bar', figsize=(25, 9), title=u'融资融券余额: (单位：万亿)')
    
    ax.set_ylim([0.8, data.tradeVal.max()])
    axhline(y=1.0, color='r')
    report.fig_style(ax, [u'万亿大关', u'两融余额'], xlabel=u'交易日')
    
    ###
    tradeValChange = [0]
    tmp = data.tradeVal[0]
    for i in data.tradeVal[1:]:
        tradeValChange.append(i - tmp)
        tmp = i
    data['tradeValChange'] = tradeValChange
    ax = data.plot(y=['tradeValChange'], kind='bar', figsize=(25, 9), title=u'当日融资融券余额相对上一交易日变动: (单位：万亿)', color='g')
    data['tmp'] = [max(i, 0) for i in data.tradeValChange]
    data.plot(y=['tmp'], kind='bar', color='r', ax=ax)
    report.fig_style(ax, [u'两融余额变动'], xlabel=u'交易日')
    
    return data

data = rong_zi_rong_quan()
````

`大宗交易`
====
***

```
大宗交易(block trading),又称为大宗买卖，是指达到规定的最低限额的证券单笔买卖申报，买卖双方经过协议达成一致并经交易所确定成交的证券交易。具体来说，各个交易所在它的交易制度中或者在它的大宗交易制度中都对大宗交易有明确的界定，而且各不相同。 
```

一般来说，大宗交易有以下一些常见的场景：

- 过桥减持
- 战略配售分仓
- 基金建仓
- 机构借助大宗交易套利
- 利益输送

````python
def dazong_jiaoyi():
    """大宗交易
    """
    trade_days = report.trade_pre_n_days(n=2)
    data = DataAPI.MktBlockdGet(secID=set_universe('A'), beginDate=trade_days[0], endDate=trade_days[-1], 
                                assetClass=u"E", field=['tradeDate', 'tradeVal'])
    data = data.groupby('tradeDate').sum()
    data.tradeVal = data.tradeVal / (10 ** 4)
    ax = data.plot(y=['tradeVal'], kind='bar', figsize=(30, 20), title=u'每日大宗交易量: (单位: 亿元)')
    report.fig_style(ax, [u'大宗交易量'], 'best', xlabel=u'交易日')
    
    return data

data = dazong_jiaoyi()
````

`大盘资金流入流出情况`
====
***

```
目前是统计各个行业都资金流入流出情况来计算大盘资金流入流出情况，可以作为宏观择时的一个参考。
```

````python
def money_flow():
    """资金流入流出情况
    """
    trade_days = report.trade_pre_n_days(n=1)
    data = pd.DataFrame()
    for day in trade_days:
        tmp = DataAPI.MktIndustryFlowOrderGet(tradeDate=day)
        data = tmp if data.empty else data.append(tmp)
    data.netInflowS, data.netInflowM = data.netInflowS / (10**9), data.netInflowM/ (10**9)
    data.netInflowL, data.netInflowXl = data.netInflowL / (10**9), data.netInflowXl/ (10**9)
    data = data.groupby('tradeDate').sum()
    ax = data.plot(y=['netInflowS', 'netInflowM', 'netInflowL', 'netInflowXl'], kind='bar', figsize=(30, 20), title=u'每日资金净流入流出情况: (单位: 十亿)')
    report.fig_style(ax, [u'小单', u'中单', u'大单', u'特大单'], 'best', xlabel=u'交易日')
    
    return data

data = money_flow()
````

`板块涨跌情况`
=====
****

```
采用 申万行业二级分类标准，计算各个行业每个交易日的平均涨幅：该行业成分股交易日涨幅的平均值；以及各个行业都涨跌家数分布：该行业成分股交易日涨跌分布；

申万行业二级分类大概有 110 个行业，目前图形中只画出平均涨幅前十，后十的部分。
```

````python
def industry_change():
    """行业涨跌平均涨幅
    """
    current_hour = dt.datetime.today().hour
    trade_days = report.trade_pre_n_days()
    preTradeDate = trade_days[-2] if current_hour < 15 else trade_days[-1]
    
    data = DataAPI.MktEqudGet(tradeDate=preTradeDate, secID=set_universe('A'), 
                              field=['secID', 'ticker', 'preClosePrice', 'closePrice', 'tradeDate'])
    
    data['up_ratio'] = (data['closePrice'] - data['preClosePrice']) / data['preClosePrice']
    data['is_up'] = [1 if i>=0 else 0 for i in data.up_ratio]
    data['is_down'] = [1 if i<0 else 0 for i in data.up_ratio]
    
    industry_ticker = DataAPI.EquIndustryGet(secID=set_universe('A'),
                                             field=['ticker', 'secID', 'secShortName', 'industryName2', 'industryName1','isNew', 'industry'])
    industry_ticker = industry_ticker[(industry_ticker.isNew == 1) & (industry_ticker.industry == '申万行业分类')]
    
    data = data.merge(industry_ticker[['ticker', 'secShortName', 'industryName2', 'industryName1']], on=['ticker'])
    
    industry2 = data.groupby('industryName2')
    
    ### plot avg ratio     
    fig = industry2.mean()
    fig = fig.sort(columns=['up_ratio'])
    tmp = fig.head(10).append(fig.tail(10))
    ax = tmp.plot(y=['up_ratio'], kind='bar', figsize=(14, 7), color='g', title=u'申万二级行业平均涨幅前十后十 : {}'.format(preTradeDate))
    tmp['tmp'] = [max(i, 0) for i in tmp.up_ratio]
    tmp.plot(y=['tmp'], kind='bar', color='r', ax=ax)
    
    for label in ax.get_xticklabels(): 
        label.set_fontproperties(font) 
    report.fig_style(ax, [u'行业平均涨幅'], 'best', xlabel=u'行业名称')
    
    ### plot up and down 
    fig = industry2.sum()
    fig['industry'] = fig.index.values
    fig = fig[fig.industry.isin(tmp.index.values)]
    fig.up_ratio = fig.up_ratio / (fig.is_up + fig.is_down)
    fig = fig.sort(columns=['up_ratio'])
    ax = fig.plot(y=['is_up', 'is_down'], kind='bar', figsize=(14, 7), 
                  stacked=True, title=u'申万二级行业涨跌情况前十后十 : {}'.format(preTradeDate), color=['r', 'g'])
    for label in ax.get_xticklabels(): 
        label.set_fontproperties(font) 
    report.fig_style(ax, [u'上涨家数', u'下跌家数'], 'best', xlabel=u'行业名称')

    return data, fig

data, fig = industry_change()
````

`行业资金流入流出`
=====
****

```
上一个交易日资金净流入最大的行业，排序方式：资金净流入最大，其次 特大单 ＋ 大单 净流入最大；
```

````python
def industry_money_flow():
    """行业资金流
    """
    current_hour = dt.datetime.today().hour
    trade_days = report.trade_pre_n_days()
    preTradeDate = trade_days[-2] if current_hour < 15 else trade_days[-1]
    data = DataAPI.MktIndustryFlowOrderGet(tradeDate=preTradeDate.replace('-', ''), 
                                           field=['industryName', 'netInflowS', 'tradeDate', 'netInflowM', 'netInflowL', 'netInflowXl'])
    tmp = DataAPI.MktIndustryFlowGet(tradeDate=preTradeDate, field=['industryName', 'moneyInflow', 'moneyOutflow', 'netMoneyInflow'])
    data = data.merge(tmp, on=['industryName'])
    
    data['zhuli'] = data.netInflowL + data.netInflowXl    
    data.sort(columns=['netMoneyInflow', 'zhuli'], ascending=[0, 0], inplace=True)
    data.netMoneyInflow, data.zhuli = data.netMoneyInflow / 10 ** 8, data.zhuli / 10 ** 8
    res = data[['industryName', u'tradeDate', 'netMoneyInflow', 'zhuli']]
    res.columns = [u'行业名称', u'交易日', u'资金净流入(单位：亿元)', u'大单，特大单净流入(单位：亿元)']
    res.reset_index(inplace=True, drop=True)
    
    return res

data = industry_money_flow()
data.head(5)
````

```
上一个交易日资金净流入最大的个股，排序方式：资金净流入最大，其次 特大单 ＋ 大单 净流入最大；
```

````python
def ticker_money_flow():
    """个股资金流
    """
    current_hour = dt.datetime.today().hour
    trade_days = report.trade_pre_n_days()
    preTradeDate = trade_days[-2] if current_hour < 15 else trade_days[-1]
    all_ticker = report.get_all_ticker(preTradeDate)
    data = DataAPI.MktEquFlowOrderGet(secID=set_universe("A"), beginDate=preTradeDate, endDate=preTradeDate, 
                                      field=['ticker', 'secID', 'tradeDate','netInflowS', 'netInflowM', 'netInflowL', 'netInflowXl'])
    tmp = DataAPI.MktEquFlowGet(tradeDate=preTradeDate, field=['ticker', 'secID', 'moneyInflow', 'moneyOutflow', 'netMoneyInflow'])
    data = data.merge(tmp, on=['secID', 'ticker'])
    
    data['zhuli'] = data.netInflowL + data.netInflowXl
    data = data.merge(all_ticker, on=['ticker', 'secID'])
    data.sort(columns=['netMoneyInflow', 'zhuli'], ascending=[0, 0], inplace=True)
    data.marketValue, data.totalShares = data.marketValue / 10 ** 8, data.totalShares / 10 ** 8
    data.netMoneyInflow, data.zhuli = data.netMoneyInflow / 10 ** 8, data.zhuli / 10 ** 8
    res = data[['ticker', 'tradeDate', 'secShortName', 'marketValue', 'totalShares', 'netMoneyInflow', 'zhuli']]
    res.columns = ['ticker', 'tradeDate', u'中文简称', u'总市值(单位：亿元)', u'总股本(单位：亿股)', u'资金净流入(单位：亿元)', u'大单，特大单净流入(单位：亿元)']
    res.reset_index(inplace=True, drop=True)
    
    return res

data = ticker_money_flow()
data.head(5)
````
