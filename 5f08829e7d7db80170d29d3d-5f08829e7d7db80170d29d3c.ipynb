```python
# 导入需要的模块
import numpy as np
import pandas  as pd
import os
import matplotlib.pyplot as plt
import scipy.stats as st
from CAL.PyCAL import *
cal = Calendar('China.SSE')
import time
import datetime
#建立文件夹路径
raw_data_path = './my_code'
if not os.path.exists(raw_data_path):
    os.mkdir(raw_data_path)    
```

```python
#定义后面需要的几个函数
def time_change(x):   #定义时间转化函数
    y = datetime.datetime.strptime(x,'%Y-%m-%d')
    z = y.strftime('%Y%m%d')
    return z
#time_change('2018-09-09')


from CAL.PyCAL import *
cal = Calendar('China.SSE')
def get_trade_list(start_time,end_time):  #定义获取所有交易日、月末、月初函数
    
    all_trade_date = DataAPI.TradeCalGet(exchangeCD=u"XSHG",beginDate=start_time,endDate=end_time,\
                                     field=u"calendarDate,isOpen,isMonthEnd",pandas="1")
    all_trade_date['calendarDate'] = map(time_change,all_trade_date['calendarDate'])
    
    trade_date = all_trade_date[all_trade_date['isOpen']==1]['calendarDate'].tolist()
    
    month_end =  all_trade_date[all_trade_date['isMonthEnd']==1]['calendarDate'].tolist()
        
    month_begin = [cal.advanceDate(date,'1B').strftime('%Y%m%d') for date in month_end]   
    
    return trade_date,month_end,month_begin    

#get_trade_list('20170101','20180901')

def st_remove(source_code,date_time = None):  #  去掉ST股票
    
    date_time = date_time if date_time is not None else datetime.datetime.now().strftime('%Y%m%d')
    
    st_secID = DataAPI.SecSTGet(beginDate=date_time,endDate=date_time,secID=source_code,\
                                field=u"tradeDate,secID",pandas="1")
    
    st_secID_list = st_secID['secID'].tolist()
    
    result_code = [code for code in source_code if code not in st_secID_list]
    
    return result_code

def ticker2secID(source_code):  # source_code  为list
    all_info = DataAPI.EquGet(equTypeCD=u"A",listStatusCD=u"L,S,DE,UN",\
                              field=u"secID,ticker",pandas="1")
    all_info.set_index('ticker',drop = True,inplace = True)
    all_info_dict = dict(all_info['secID'])
    
    result_code = []
    
    if isinstance(source_code,list):
        for code in source_code:
            if code in all_info_dict.keys():
                result_code.append(all_info_dict[code])
            else:
                print(code+'不存在')        
    else:
        print('请输入list')
            
    return result_code

#ticker2secID(['600763','002060'])


import datetime
def new_remove(source_ticker,N = 20,date_time = None):
    
    date_time = date_time if date_time is not None else datetime.datetime.now().strftime('%Y%m%d')
    period = str((-1)*N)+'B'
    new_time = cal.advanceDate(date_time,period).strftime('%Y-%m-%d')
    code_ipo = DataAPI.SecIDGet(assetClass=u"e",ticker = source_ticker,field=u"ticker,listDate",pandas="1")
    #code_ipo['listDate']=map(time_change,code_ipo['listDate'])
    code_ipo_new = code_ipo[code_ipo['listDate']>new_time]['ticker'].tolist()
    
    result_code = [code for code in source_ticker if code not in code_ipo_new]
    
    return result_code
    
def get_pool_code(month_begin,days=180):
    
    univ = DynamicUniverse('A') 
        
    month_code = univ.preview(month_begin,skip_halted = False)  #确定是否去掉停牌
    
    month_code = st_remove(month_code,date_time = month_begin) 
    month_code = [x.split('.')[0] for x in month_code]
    month_code = new_remove(month_code,N =days,date_time = month_begin)
    month_code = ticker2secID(month_code)
    
    month_code_dict = dict({'code':month_code})
    
    month_code_pd = pd.DataFrame(month_code_dict)
    month_code_pd['date'] = month_begin
    month_code_pd = month_code_pd[['date','code']]
    
    return  month_code_pd

#get_pool_code('20200701',days=180)
```

```python
start_time = '20071220'
end_time = '20090601'
trade_date,month_end,month_begin = get_trade_list(start_time,end_time)

all_stock_pd= pd.DataFrame()

for date in month_begin:    
    
    month_code_pd=get_pool_code(date,days=180)    
    
    all_stock_pd = all_stock_pd.append(month_code_pd)
    
all_stock_pd.head()  #  每个月月初组成的股票池    
```

```python
# 获得所有交易日，设计的所有股票的交易信息
all_code = sorted(list(set(list(all_stock_pd['code']))))
market_info_frame = DataAPI.MktEqudGet(secID=all_code,beginDate='20071220',endDate='20090601',isOpen=1,\
                                       field=u"tradeDate,secID,turnoverRate,chgPct",pandas="1")
#改变研究区间的时候，需要改变上面的日期
market_info_frame['tradeDate'] = map(time_change,market_info_frame['tradeDate'])
market_info_frame =market_info_frame[['tradeDate','secID','turnoverRate','chgPct']]
market_info_frame.columns = ['date','code','to','ret']
market_info_frame = market_info_frame.dropna()
market_info_frame.head()
```

```python
# 以每一个月月底为一个节点，分别计算三个因子fr, frd ,fru
all_signal = pd.DataFrame()

for i in range(1,len(month_end)):
    
    #定义四个时间点
    begin = month_begin[i-1]
    end =  month_end[i]
    start = month_begin[i]
    
    if i < len(month_end)-1:
        next_end = month_end[i+1]
    else:
        next_end ='20090630'  # 改变研究区间的时候需要改变
    
    code = all_stock_pd[all_stock_pd['date']==start]['code'].tolist()  #下一个月月初的娥股票代码，为研究该月的代码标准
    
    #获取所有股票下一个月的收益
    info1 = DataAPI.MktEqudAdjAfGet(secID=code,tradeDate=end,\
                                    field=u"tradeDate,secID,closePrice",pandas="1")
    info1 = info1[['tradeDate','secID','closePrice']]
    info1.columns  = ['tradeDate1','code','closePrice1'] 
    
    info2 = DataAPI.MktEqudAdjAfGet(secID=code,tradeDate=next_end,\
                                  field=u"tradeDate,secID,closePrice",pandas="1")
    info2 = info2[['tradeDate','secID','closePrice']]
    info2.columns  = ['tradeDate2','code','closePrice2']  
    
    info  = pd.merge(info1,info2,on = 'code',how = 'outer')
    info['month_ret'] = info['closePrice2']/info['closePrice1']-1
    info = info[['code','month_ret']]
    info['date'] = end
    info =info[['date','code','month_ret']]
    
    # 做事情的效率，效果，明确地目标。想法设法的高效。只运行2年看效果，一步一测试
    
    # 获取三个因子
    period_frame = market_info_frame[(market_info_frame['date']>=begin)&(market_info_frame['date']<=end)]
    period_frame = period_frame[period_frame['code'].isin(code)]
    period_frame = period_frame.sort_values(by = ['code','date'],axis = 0)
    
    
    period_frame['next_ret'] = period_frame.groupby('code')['ret'].shift(-1,axis = 0)
    period_frame['abs_next_ret'] = abs(period_frame['next_ret'])
    
    
    #获取FR因子
    #num_days = [x for x in trade_date if x>=begin and x <=end]  这一步省略，节约时间  min_periods = len(num_days)/2
    signal_fr=period_frame.groupby('code').apply(lambda df: df['to'].corr(df['abs_next_ret'],method = 'pearson'))     
    # 上面这一步，必须独立生成新的series,每一个股票，每一个月末都会产生一个新的FR
    
    
    #获取FRu因子
    period_frame_positive = period_frame[period_frame['next_ret']>=0]
    signal_fru=period_frame_positive.groupby('code').apply(lambda df: df['to'].corr(df['next_ret'],method = 'pearson')) 
    
    #获取FRu因子
    period_frame_negative = period_frame[period_frame['next_ret']<0]
    signal_frd = period_frame_negative.groupby('code').apply(lambda df: df['to'].corr(df['next_ret'],method = 'pearson'))
    
    #将三个因子合并到一个DF
    signal_month = pd.concat([signal_fr,signal_fru],axis = 1)
    signal_month.columns = ['fr','fru']
    signal_month = pd.concat([signal_month,signal_frd],axis = 1)
    signal_month.columns = ['fr','fru','frd']
    
    signal_month['code'] = signal_month.index
    signal_month.reset_index(drop = True,inplace = True)
    
    # 与前面得到的下一个月的收益率结合
    signal_month = pd.merge(info,signal_month, on='code',how = 'outer')
    # 数据经常出现对不上，这是正常现象，需要你去按照有个标准去统一。
    
    all_signal = all_signal.append(signal_month)
    
all_signal
```

```python
all_signal.to_csv('my_code/all_signal.csv',index = True)
```

```python
# 定义获取IC值的函数
def cal_ic(signal_name,signal_df):
    ic_series = signal_df.groupby('date').apply(lambda df:df[signal_name].corr(df['month_ret'],method = 'spearman'))
    ic_df = pd.DataFrame(ic_series)
    ic_df.columns = [signal_name]
    return ic_df


ic_fr_df = cal_ic('fr',all_signal)
ic_fru_df = cal_ic('fru',all_signal)
ic_frd_df = cal_ic('frd',all_signal)

all_signal_ic = pd.merge(ic_fr_df,ic_fru_df,left_index = True,right_index = True)
all_signal_ic = pd.merge(all_signal_ic,ic_frd_df,left_index = True,right_index = True)
all_signal_ic
```

```python
#计算此时计算出来的所有的因子的IC ICIR值
fr_ic_icir_df = pd.DataFrame(index = ['fr','fru','frd'],columns=['IC','ICIR'])
for signal_name in ['fr','fru','frd']:
    fr_ic_icir_df.loc[signal_name,'IC'] = all_signal_ic[signal_name].mean()
    fr_ic_icir_df.loc[signal_name,'ICIR'] = all_signal_ic[signal_name].mean()/\
    all_signal_ic[signal_name].std()*np.sqrt(12)
    
fr_ic_icir_df
```

```python
#计算三个纯净因子
import statsmodels.api as sm
else_factors = pd.DataFrame()

#定义获取行业的函数
def get_industry(date,code):
    code_indu = DataAPI.MdSwBackGet(secID=code,field=u"secID,isNew,industryName1",pandas="1")
    code_indu = code_indu[code_indu['isNew']==1]
    code_indu['tradeDate'] = date
    
    code_indu = code_indu[['secID','industryName1','tradeDate']]
    code_indu.columns = ['code','industryName','date']
    return code_indu
    

all_pure_fr = pd.DataFrame()    
    
for i in range(1,len(month_end)):
    
    #定义四个时间点
    begin = month_begin[i-1]
    end =  month_end[i]
    start = month_begin[i]
    
    if i < len(month_end)-1:
        next_end = month_end[i+1]
    else:
        next_end ='20090630'  # 改变研究区间的时候需要改变
    
    code = all_stock_pd[all_stock_pd['date']==start]['code'].tolist()  #下一个月月初的娥股票代码，为研究该月的代码标准
    
    #获取行业信息
    code_indu = get_industry(end,code)
    
    #获取其它的因子
    factors = DataAPI.MktStockFactorsOneDayGet(tradeDate=end,secID=code,field=u"secID,tradeDate,LFLO,REVS20,VOL20,ATR14",pandas="1")
    factors['tradeDate'] = map(time_change,factors['tradeDate'])
    factors =factors[['secID','tradeDate','LFLO','REVS20','VOL20','ATR14']]
    factors.columns = ['code','date','LFLO','REVS20','VOL20','ATR14']

    # 获取原先构造的是三个因子
    fr_frame = all_signal[all_signal['date'] == end]
    
    
    #将行业和其它因子，以及三个因子结合到一起
    factors = pd.merge(factors,code_indu,on=['code','date'],how ='outer')
    factors = pd.merge(factors,fr_frame,on=['code','date'],how ='outer')
    
    factors = factors[['code','date','LFLO','REVS20','VOL20','ATR14','industryName','month_ret','fr','fru','frd']]
    factors =factors[['date','code','month_ret','fr','fru','frd','LFLO','REVS20','VOL20','ATR14','industryName']]
    
    factors.index = factors['code']
    factors = factors.drop('code',axis =1)
    
    factors.dropna(inplace = True) #进行标准化以及极值化时不可以有NA
    
    factors_columns = list(factors.columns)
    
    for col in factors_columns[2:-1]:
        temp = standardize(winsorize(factors[col]))
        factors[col] = temp
    #
    
    #将行业变为哑变量
    indu_dumm = pd.get_dummies(factors['industryName'])
    factors = factors.drop('industryName',axis =1)
    factors = pd.merge(factors,indu_dumm,left_index = True,right_index = True,how = 'outer')
    
    
    #回测构造新的因子变量
    
    y1 = np.array(factors['fr'])
    y2 = np.array(factors['fru'])
    y3 = np.array(factors['frd'])
    
    x = np.array(factors[factors.columns[5:]])
    x = sm.add_constant(x,has_constant = 'add')    
    
    #prue_fr
    model = sm.OLS(y1,x,missing = 'drop')
    result = model.fit()
    factors['fr'] = result.resid
    factors['fr'] = standardize(winsorize(factors['fr']))
    
    #prue_fru
    model = sm.OLS(y2,x,missing = 'drop')
    result = model.fit()
    factors['fru'] = result.resid
    factors['fru'] = standardize(winsorize(factors['fru']))    
    
    #prue_frd
    model = sm.OLS(y3,x,missing = 'drop')
    result = model.fit()
    factors['frd'] = result.resid
    factors['frd'] = standardize(winsorize(factors['frd']))
    
    factors['code'] = factors.index
    factors.reset_index(drop = True,inplace = True)
    
    factors = factors[['code','date','month_ret','fr','fru','frd']]
    
    all_pure_fr = all_pure_fr.append(factors)
    
all_pure_fr
```
