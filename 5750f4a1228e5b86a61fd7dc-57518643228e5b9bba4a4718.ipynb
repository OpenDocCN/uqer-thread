# Zura论文附录中WorldQuant的Alpha#03

**Alpha#03**
$$(-1 * correlation(rank(open), rank(volume), 10))$$

先了解下两个运算符：<br>
**Rank(x)** ： 排列所有工具的 x 值，然后返回的数值是介于0.0和1.0之间均匀分布的浮点数。<br>
例如假设你有一个向量：<br>
A = [29, 12, 5, 7, 10]<br>
现在，将会为向量里的每个元素**按升序分级**，最小一级为 0：<br>
A = [29, 12, 5, 7, 10]<br>
A' = [4, 3, 0, 1, 2] <br>
每个“等级”再除以（元素个数 - 1）。于是得到：<br>
Rank (A) = [1.00, 0.75, 0.00, 0.25, 0.50]<br>
所以您等到了一个从 0 到 1 间距相等的向量。<br>

**Correlation(x, y, n)** : 	计算向量 x 中的值在过去n天的相关系数。请注意，n 必须小于256。

那么，该公式的意思就很明朗了：首先对过去10天的openPrice和turnoverVol进行Rank操作，然后计算两个向量的相关系数。以相关系数的倒数计算权重对选定的股票进行调仓。<br>

该式的关键在openPrice 和 成交量负相关，即价格减小时，成交量增加；价格增加时，成交量减小。这是典型的价量背离现象：<br>

**1.量增价跌**<br>
量增价跌主要是指个股（或大盘）在成交量增加的情况下、个股的股价却反而下跌的一种量价配合现象。量增价跌现象大部分是出现在下跌行情的初期，也有小部分是出现在上升行情的初期。不过，量增价跌的现象在上升行情和下降行情中的研判也是不一样的。
在上升行情初期，有的股票也会出现量增价跌现象。当股价经过一段比较长时间的下跌和底部较长时间盘整后，主力为了获取更多的低位筹码，采取边打压股价边吸货的手段，造成股价走势出现量增价跌现象，但这种现象也会随着买盘的逐渐增多、成交量的同步上扬而消失，这种量增价跌现象是底部买入信号。<br>
量增价跌的原因：<br>
（1）熊市逃命。主力打压，散户空杀空，跑出来就是“胜利大逃亡”。如股指08年5月7日。所以会放量下跌。<br>
（2）大盘调整。市场信心不足，抛盘就会加重，主力为了稳定股价，必然在低位承接。如股指09年5月19日。所以会放量下跌。<br>
（3）主力派发后期。主力已经基本把股票派发完毕，获利丰厚。但还余少量股票，减价处理就是，已经不在乎股价的问题了，但因为筹码有限，成交量只是相对放大。所以会放量下跌。<br>
（4）机构调仓。就是放弃一个股票，腾挪出资金再另建新仓。如歌华有线09年1月23日。所以会放量下跌。<br>
（5）高配送后减仓。高配送后，经过填权，主力为了变现，会有一定的减仓行为。但因为是减仓，所以下跌的幅度不会很大，而且也不存在持续性。所以会放量下跌。<br>
（6）波浪顶。股票的走行都是呈连续的波浪状态。在波浪顶部，主力做部分的派发，再于低位承接，由于存在差价，资金可以得到增加，而持有的股票数量不变。（或资金不变，股票数量增加。）所以会放量下跌。<br>
（7）震仓。有的时候会放量下跌，造成主力派发的假象。但持续的时间都不会很长。<br>
补充一下，量价背离也就是线（均量线）价背离，也就是[麦氏理论](http://baike.baidu.com/link?url=GWnJ3cLSJw5P2JbLthgivFixV7vnpaw8UyoasBOy2BRlPLoeXxY0blMo5blne3x4wuQ_uRz7rVgXf7nXhQkXOK)中的，不合理、和不协调。但也不能一概而论。



**2.量缩价涨**<br>
量缩价涨主要是指个股（或大盘）在成交量减少的情况下、个股的股价却反而上涨的一种量价配合现象。量缩价涨多出现在上升行情的末期，有一小部分也会出现在下降行情中期的反弹过程中。<br>
量缩价涨的主要原因:<br>
（1）现在的低换手率（就是成交量萎缩），是用前期的高换手率或交易时间换来的。<br>
（2）主力控盘度高。解释：看看和讯网站的筹码分析，在中国的股票市场，50%的筹码归主力所有（当然，主力也分三六九等）。50%的筹码归散户所有。所以，即使散户抛压，问题也不大，只要主力稍作承接，股价就可以缩量上涨。<br>
（3）大盘带动。解释：在某一天，大盘上涨，可主力并没有准备在这一天有所动作，因为大盘带动，所以市场购买踊跃，散户多对多。但因为买不到太多的股票，所以缩量上涨。<br>
（4）散户惜售。解释：在牛市，很多人都希望自己的股票一涨再涨，所以，主力一拉升，没有抛盘，自然就缩量上涨。<br>
（5）场外资金或空方观望。解释：没有新资金介入或没有大的抛压，就不可有大的成交量出现。所以缩量上涨。<br>
（6）套牢盘太重。解释：分两种情况，在熊市时，在中国的股票市场，有很多的人，已经把自己的本金抽出，单独用盈利的资金来炒作股票。所以，他就会想，套着呗，反正现在的钱也是以前赚的，等有一天涨回来，到时候，我还得连本带利地赚回来。即使是用自己的本金买股票的人，有此想法的也大有人在。另外，在普通股民当中，止损的概念还很模糊；在牛市，就更不用说了，更看到希望了，我都套了这么长时间了，还忍不住啊。所以股市里有以量破价之说，而且此现象的发生也屡见不鲜。就是用很多的钱，去买套牢盘，以使股价突破并继续上升。所以，以量破价的道理也没什么奥秘可言。<br>
（7）诱多。判断诱多有三个条件：1.短线均量线下跌；2.当日的成交量稍高于60日或120日均线；3.当天该股票价格上涨。<br>
