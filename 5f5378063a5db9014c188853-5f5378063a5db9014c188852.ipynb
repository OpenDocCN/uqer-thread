# ETF波动率观察

```python
#今天被芯片ETF 512760 还有中国软件 玩了两次过山车，要死要活的
#找点方法看看是否短期过热，逃顶，可以用在板块or个股上判断是否进入加速冲刺赶顶的阶段~
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
ss='20200101'
ee='20200905'
etf='512760.XSHG'
f=['secID','tradeDate','closePrice','circulationShares','turnoverValue']
rawdata=DataAPI.MktFunddGet(secID=etf,ticker=u"",tradeDate=u"",\
                         beginDate=ss,endDate=ee,field=f,pandas="1")
closePrice=rawdata['closePrice']
days=rawdata['tradeDate']
turnoverValue=rawdata['turnoverValue']
#计算对数收益率
# print(closePrice)
closePrice=np.log(closePrice)
# print(closePrice)
tmp=closePrice.shift(1)
# print(tmp)
logReturn=closePrice-tmp
del tmp
# print(logReturn)
out=pd.concat([days,logReturn],axis=1)
# print(out)
out=out[1:]#因为第一个是NaN
out=out.reset_index(drop=True)
turnoverValue=turnoverValue[1:]
# print(out)
windowsize=14



record_std={'tradedate':[],'STD':[],'ts':[]}
for i in range (windowsize-1,len(logReturn)-1):
    # print(out['closePrice'][i])
    tmp=out['closePrice'][i-13:i+1]
    day=out['tradeDate'][i]
    # print(day)
    record_std.get('tradedate').append(day)
    test=np.std(tmp)*np.sqrt(360)#用年化来放大区别，方便观察
    record_std.get('STD').append(test)
    record_std.get('ts').append(turnoverValue[i])



    
```

```python
record_std=pd.DataFrame(record_std)
import datetime
record_std['tradedate']=pd.to_datetime(record_std['tradedate'])
print(record_std.dtypes)
# record_std=pd.DataFrame(record_std)
fig=plt.figure(figsize=(15,10))

print(len(record_std))
ax1=fig.add_subplot(111)
ax1.plot(record_std['tradedate'],record_std['STD'])
ax2 = ax1.twinx()
# print(record_std['ts'])
# print(record_std['STD'])
ax2.plot(record_std['tradedate'],record_std['ts'],'r')
plt.show()
```
