# 如何正确把玩 DataAPI Cache

## `介绍`

首先看看 `优矿升级日志`: https://uqer.datayes.com/help/changelog/ , 发现 DataAPI 新增了一个 cache 的功能：https://uqer.datayes.com/help/faqApi/#DataAPI缓存加速

您可以开启DataAPI缓存加速开关，如果多次调用同一个API的相同数据，该数据将会缓存在您云端研究环境的内存中，提升访问效率。

该开关默认关闭，具体使用方法例如：

```
# 默认为False，不开启缓存
DataAPI.settings.cache_enabled = False
# 开启缓存，当前Notebook所有DataAPI数据都会缓存
DataAPI.settings.cache_enabled = True
```

## `结论`

- 调用100次 `DataAPI.SecIDGet`
    + 无cache 时：17.41 秒
    + 有cache 时：0.19 秒
- 策略：[『小试牛刀』个股资金流](https://uqer.datayes.com/community/share/5696099e228e5b18dfba2c8b)，回测时间可以：回测详情 -> 配置
    + 无cache 时回测时间：5*60 ＝ 300 秒
    + 有cache 时回测时间：3 秒

## `适用场景`

- 开启 cache 后，能大幅提升回测速度
- 开启 cache 会导致 开始研究资源 －> 内存 使用率较高，不过如果只是调优一个策略的话，1g 左右的内存能满足大多数 cache 需求

## `把玩攻略`

- 在策略调参时使用 dataapi cache
- 在策略研究时使用 dataapi cache
- 提交模拟交易或参加比赛时关掉 dataapi cache

````python
# built-in
import time
import operator
import collections
from datetime import datetime
from functools import wraps

# 3-party
from CAL.PyCAL import *
import pandas as pd
pd.options.display.max_columns = 100
pd.options.display.max_rows = 300
import numpy as np
import seaborn
from matplotlib import pylab
import matplotlib.pyplot as plt
````

````python
from functools import wraps

def wrapper_func_time(func):
    @wraps(func)
    def wrapped(*args, **kwargs):
        tstart = time.time()
        res = func(*args, **kwargs)
        tend = time.time()
        
        print '-' * 80
        print 'consume {} seconds for func {}'.format(tend - tstart, func.func_name)
        return res
    return wrapped


############################################
### constants
############################################

def const_trade_calendar():
    data = DataAPI.TradeCalGet(exchangeCD=u"XSHG,XSHE", beginDate=u"20050101", endDate=u"", field=['calendarDate', 'isOpen'])
    data = data[data.isOpen == 1]

    return list(data.calendarDate)


def const_pre_trade_day(day):
    trade_days = const_trade_calendar() 
    index = trade_days.index(day)
    
    return trade_days[index - 1]

def get_all_ticker(tradeDate, preTradeDate, sector=['主板', '中小板', '创业板']):    
    tmp = DataAPI.EquGet(equTypeCD="A", field=['ticker', 'secID', "listDate", "ListSector", "totalShares"])
    tmp = tmp[tmp.ListSector.isin(sector)]
    equ_pool = DataAPI.MktEqudGet(tradeDate=preTradeDate, 
                                  field=["ticker", "exchangeCD", "secShortName","preClosePrice", "openPrice", "dealAmount", 
                                         "highestPrice", "lowestPrice", "closePrice", "negMarketValue", "marketValue", "isOpen"])
    tmp = pd.merge(left=equ_pool, right=tmp, left_on="ticker", right_on="ticker")
    return tmp
````

````python
@wrapper_func_time
def data_api_test_cache(n=100):
    DataAPI.settings.cache_enabled = True
    for i in range(n):
        _ = DataAPI.SecIDGet(partyID=u"",assetClass=u"E",ticker=[],cnSpell=u"",field=u"",pandas="1")
    return None

@wrapper_func_time
def data_api_test_no_cache(n=100):
    DataAPI.settings.cache_enabled = False
    for i in range(n):
        _ = DataAPI.SecIDGet(partyID=u"",assetClass=u"E",ticker=[],cnSpell=u"",field=u"",pandas="1")
    return None
````
