```python
#coding=utf-8
import mercury
import pandas as pd
import datetime,time
if __name__ == "__main__":
    uqername='*********'
    uqerpwd='********'
    taskdate='2000-01-01'
    todaytask=False
    while True:
        lastdate = datetime.datetime.now() + datetime.timedelta(days=-1)  # 昨天
        lastdate = lastdate.strftime('%Y-%m-%d')

        if taskdate <> lastdate:
            taskdate = lastdate
            todaytask = False
        while todaytask == False:
            filename = lastdate + '.csv'
            try:
                #执行取文件
                uqer = mercury.Client(uqername, uqerpwd)
                # all_files = uqer.list_data()
                uqer.download_data(filename)
                time.sleep(10)
                try:
                    # 打开文件尝试是否下载成功
                    pd.read_csv(filename)
                    todaytask = True
                    print 'open ' + filename + ' successful.' + datetime.datetime.now().strftime('%Y-%m-%d %H:%M:%S')
                    # 执行任务。。
                    #domyproduce()
                except:
                    todaytask = False
                    print 'open ' + filename + ' failed.'
            except:
                print '.not find file ' + datetime.datetime.now().strftime('%Y-%m-%d %H:%M:%S')
                time.sleep(3600)
        time.sleep(3600) #3600 1hours
    #task end while
```
