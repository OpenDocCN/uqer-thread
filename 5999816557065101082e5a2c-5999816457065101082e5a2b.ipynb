##四、statsmodels介绍
最后来介绍statsmodels数据库，Statsmodels是Python的统计建模和计量经济学工具包，包括一些描述统计、统计模型估计和推断。


```python
import statsmodels.api as sm
 
# Generate artificial data (2 regressors + constant)
nobs = 22
X = data["closePrice"]
X = sm.add_constant(X)
beta = [1, .5]
e = np.random.random(nobs)
y = np.dot(X, beta) + e
 
# Fit regression model
results = sm.OLS(y, X).fit()
 
# Inspect the results
print(results.summary())
```
