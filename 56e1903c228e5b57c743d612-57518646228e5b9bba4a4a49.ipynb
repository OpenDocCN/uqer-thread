# 大小盘指数对比

```python
import numpy as np
import datetime as dt
import seaborn as sns
from pandas import DataFrame
from matplotlib import pyplot as plt
from CAL.PyCAL import *

zhishu = [('000043', 'huge'), ('000044','mid'), ('000045','small'), ('000046', 'mid&small'), ('000047', 'quanzhi')]

begind = '20140101'
endd = '20160310'
fd =  DataAPI.MktIdxdGet(ticker='000001',beginDate=begind,endDate=endd,field='closeIndex,tradeDate')
fd.columns = ['shangzheng','tradeDate']
fd.index = fd.tradeDate.apply(lambda x: dt.datetime.strptime(x, '%Y-%m-%d'))

for tk,name in zhishu:
    df = DataAPI.MktIdxdGet(ticker=tk,beginDate=begind,endDate=endd,field='closeIndex')
    fd[name] = np.NAN
    fd[name][:] = df.closeIndex
    
fd.plot(figsize=(20,15))

```
