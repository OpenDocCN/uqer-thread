# 二月日期效应

```python
def timing_February(context):
    """
    二月买入全部的中小板指股票持有一个月到三月卖出
    """
    # 查看账户等仓位信息
    today = context.current_date  # 当前日期
    account = context.get_account("fantasy_account")
    current_universe = context.get_universe(exclude_halt=True)
    security_position = account.get_positions()
    cash = account.cash
    
    buy_list = []  #需要买入的股票代码
    sell_list= []  #需要卖出入的股票代码
    
    for sec in current_universe:# 查看持仓信息
        #。查看当先是否是当前2月第一个交易日买入 
        # 二月第二个交易日不符合第二个条件（即是否持有股票），所以只买入一次
        if today.month==2 and sec not in security_position: 
            buy_list.append(sec)
        elif today.month==3 and sec in security_position: # 三月卖出
            sell_list.append(sec)
    
    
    # 择时策略部分
    m = len(sell_list)
    for sec in sell_list[:m]:
        account.order_to(sec, 0)   
    d = min(len(buy_list), int(cash) // 100000)
    
    for sec in buy_list[:d]:
        account.order(sec, 100000/ context.current_price(sec))
        
        
    
        
```

```python
def timing(context):
    timing_February(context)
  
    
```
