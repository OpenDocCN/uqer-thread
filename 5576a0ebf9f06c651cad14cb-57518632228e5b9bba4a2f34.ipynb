# 站在风口上带你飞

> 2014年以来，市场上流行这样一句话：“站在风口上，猪都能飞起来。”其中的“风”指的大体上就是消息面和热度。本篇中，我们从板块热度轮动的角度，构建**板块热点投资策略**，对牛市中由于投机氛围浓郁带来的板块轮动现象进行探讨。

1. 策略逻辑
------------------------
本篇中，我们选取沪深300行业分类作为研究对象：000908.XSHG(300能源)，000909.XSHG(300材料)，000910.XSHG(300工业)，000911.XSHG(300可选)，000912.XSHG(300消费)，000913.XSHG(300医药)，000914.XSHG(300金融)，000915.XSHG(300信息)，000916.XSHG(300电信)，000917.XSHG(300公用)。

我们的策略流程是：
- 通过板块成份股构建每一板块的**流通市值加权板块热度指数**
- 买入前期平均板块热度最高的板块中，加权平均热度最高的5只股票

即**买入热点板块中的热点股票**。

在我们的**量化实验室**中，用户可以通过数据API: ``MktTickRTSnapshotIndexGet`` 获取某板块对应的**板块成份股列表**。通过``NewsHeatIndexGet`` 获取所有成份股对应的**新闻热度信息**。通过``MktEqudGet`` 获取所有成份股对应的**流通市值信息**。

```python
import pandas as pd
import numpy as np

#ticker转换为id
tk2id=lambda x: x+'.XSHG' if x[0]=='6'else x+'.XSHE'

# 所有板块的成分股录入
def importAllSectors(all_sector):
    sector_stock = {}
    for sector in all_sector:
        stock = DataAPI.MktTickRTSnapshotIndexGet(securityID=sector,field=u"prevClosePrice,openPrice",pandas="1")
        # 获取该板块所有成分股信息
        stock_code = map(tk2id,stock['ticker'])
        sector_stock[sector] = stock_code
    return sector_stock

# 所有板块在一段时间内板块热度字典（不带权重与带权重）
def sectorHeat(start, end, sector_stock):
    sector_heat = {}
    sector_heatWithValue = {}
    sector_info = {}
    
    for sector in sector_stock:
        stock_code = sector_stock[sector] 
        # 获取所有成份股的新闻热度信息
        news_heat = DataAPI.NewsHeatIndexGet(secID = stock_code, beginDate = start.strftime('%Y%m%d'), endDate = end.strftime('%Y%m%d'),field=['secID', 'newsPublishDate', 'heatIndex'])
        
        # 获取所有成份股的流通市值信息
        total_capital = DataAPI.MktEqudGet(secID= stock_code, beginDate = start.strftime('%Y%m%d'), endDate = end.strftime('%Y%m%d'),field=['secID', 'tradeDate', 'negMarketValue'])
        
        # 获取以市值加权的新闻热度信息
        heat_capital = pd.merge(news_heat, total_capital, left_on = ['secID','newsPublishDate'], right_on = ['secID','tradeDate'])
        heat_capital['heatWithValue'] = heat_capital['heatIndex'] * heat_capital['negMarketValue'] / 10000000000
        
        # 获取板块热度
        heat_capital_date = heat_capital.groupby('newsPublishDate')
        heatIndex = heat_capital_date.sum()
        
        # 标准化数据
        heatIndexNorm = heatIndex['heatIndex']
        heatIndexNorm = (heatIndexNorm-heatIndexNorm.mean())/heatIndexNorm.std()
        heatIndex['heatIndexNorm'] = heatIndexNorm
        heatIndexWithValueNorm =heatIndex['heatWithValue']
        heatIndexWithValueNorm = (heatIndexWithValueNorm-heatIndexWithValueNorm.mean())/heatIndexWithValueNorm.std()
        heatIndex['heatIndexWithValueNorm'] = heatIndexWithValueNorm

        # 板块热度导入sector_heat
        sector_info[sector] = heat_capital
        sector_heat[sector] = heatIndex['heatIndexNorm']
        sector_heatWithValue[sector] = heatIndex['heatIndexWithValueNorm']
        
    data_sector_heat = pd.DataFrame(sector_heat)
    data_sector_heatWithValue = pd.DataFrame(sector_heatWithValue)
    return data_sector_heat, data_sector_heatWithValue, sector_info

# 板块一段时间内平均热度TOPN
def sectorHeatTop(data_sector_heat, all_sector, n):
    temp = []
    for sector in all_sector:
        temp_mean = data_sector_heat[sector].mean()
        temp.append(temp_mean)
    temp_sector = np.array(all_sector)
    # 选出前10天平均行业热度前N的行业
    sector_choose_topn = list(temp_sector[np.argsort(temp)][10-n:10])
    return sector_choose_topn

# 选出板块中热度TOPN的股票
def stockHeatTop(sector_choose_topn,sector_info, n):
    stockList = []
    for sector in sector_choose_topn:
        res = sector_info[sector]
        res = res.groupby('secID')
        topN = res.mean().sort('heatWithValue', ascending=False).head(n)
        stockList = stockList + list(topN.index)
    return stockList


        
```
