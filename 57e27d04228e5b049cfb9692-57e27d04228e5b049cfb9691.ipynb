```for stock in list(getData_current_date.index):```for循环的条件保证了在回测期间每个交易日，更够利用```stock```变量遍历```getData_current_date```中所有的证券代码
```if((getData_current_date['MA5'][stock]>getData_current_date['MA10'][stock])&(getData_last_date['MA5'][stock]<=getData_last_date['MA10'][stock])):```是“5日均线上穿10日均线”的核心代码，原作模仿了“牛顿二分法”求方程解的思想：如果某个证券在```current_date```这个交易日5日均线上穿了10日均线，那么自然在这一天5日均线的值大于10日均线的值，也就是```getData_current_date['MA5'][stock]>getData_current_date['MA10'][stock]```，同时，前一个交易日，即```last_date```这天，5日均线的值应小于10日均线的值，也就是```getData_last_date['MA5'][stock]<=getData_last_date['MA10'][stock]```。在满足了这两个条件之后，```buylist```这个之前初始化的空列表将会通过```append()```方法添加这只股票的证券代码。
将以上的判定标准取反，就可以应用到```selist```等待卖出的证券代码列表。

````python
    for stock in selist:
        if(stock in account.valid_secpos):
            order_to(stock,0)
        else:
            pass
        
    for stock in buylist:
        if(stock in account.valid_secpos):
            pass
        else:
            if(len(buylist)>=5):
                order(stock,account.cash/account.referencePrice[stock]/len(buylist))
            else:
                order(stock,account.cash/account.referencePrice[stock]/5)
            
    print "日期：",account.current_date,",持仓：",account.valid_secpos
````

胜利在望！剩下的代码就比较简单了。

```for stock in selist:```这个循环首先是证券要在```selist```当中，同时满足要在```account.valid_secpos```中，也就是quartz框架下```account```可交易证券的范围内存在，才可以卖出，卖出代码为```order_to(stock,0)```,表示一次性全部卖出。
```for stock in buylist:```买入列表原理相似。首先证券代码要在```buylist```当中，但如果证券已经在```account.valid_secpos```中，那么什么都不做，```pass```。如果不在```account.valid_secpos```中，就可以买入了。与```selist```一次性全部卖出不同，这里对是n种股票进行平均持仓。这里我倒认为原作```if...else```判断语句并没有太大必要，可以直接写```order(stock, account.cash/account.referencePrice[stock]/len(buylist))```，```account.cash```表示```account```中持有的现金数量，```account.referencePrice[stock]```是证券代码为```stock```的股票在上一个K线结束的收盘价，```len(buylist)```返回```buylist```列表中元素的个数。关于```account.cash```、```account.referencePrice[stock]```以及```order_to()```、```order()```等信息，可以参考[常用API文档](https://uqer.datayes.com/help/faqApi/)中的相关内容。
