# 逻辑回归优矿实现（优矿因子）

在网上扒了一下机器学习的例子，在优矿实现一下，看看效果。随机选取优矿因子库中的几个因子，用逻辑回归方法建模：
1、随机取一天，取period日前的因子数据和当天相对于period日前的收益率

```python
import matplotlib.pyplot as plt
from matplotlib import rc
from matplotlib import dates
rc('mathtext', default='regular')
import seaborn as sns
sns.set_style('white')
import datetime
import numpy as np
import pandas as pd
import time
import scipy.stats as st
from CAL.PyCAL import *    # CAL.PyCAL中包含font
cal = Calendar('China.SSE')

factors=['secID','tradeDate','ROE','PS','PB','ILLIQUIDITY']
period = 20

tradeDate = '2013-10-09'
last_date = cal.advanceDate(tradeDate,'-'+str(period)+'B',BizDayConvention.Preceding).strftime("%Y%m%d")

df_price = DataAPI.MktEqudGet(tradeDate=u"",secID=u"",ticker=u"",beginDate=last_date,endDate=tradeDate,isOpen="1",field=u"secID,tradeDate,closePrice",pandas="1").replace([-np.inf, np.inf], 0).fillna(value=0)
df_price['pct_close'] = df_price.groupby('secID').pct_change(period)
df_price = df_price[df_price['tradeDate']==tradeDate].replace([-np.inf, np.inf], 0).fillna(value=0)
print df_price.head(5)

df_factor = DataAPI.MktStockFactorsOneDayGet(tradeDate=last_date,secID=u"",ticker=u"",field=factors,pandas="1").replace([-np.inf, np.inf], 0).fillna(value=0)
df_factor.head(5)
```

2、拼接一下

```python
result = pd.merge(df_factor, df_price, how='inner', on=['secID'])
result.head(5)
```

3、把因子x和收益率y弄出来，并对因子集做标准化和sigmoid

```python
import lib.gra as gra

filter_column = ['secID', 'tradeDate_x','tradeDate_y','closePrice','pct_close']
result = gra.standard(result,filter_column)
result = gra.sigmoid(result,filter_column)
result_x = result.drop(['secID', 'tradeDate_x','tradeDate_y','closePrice','pct_close'], axis=1)
result_y = result['pct_close']
```

4、逻辑回归一下

```python
m, n = np.shape(result_x)  
n_y = np.shape(result_y)
print m,n,n_y

numIterations =10000  
alpha = 0.0005  
theta = np.ones(n)  
theta = gra.gradientDescent(result_x, result_y, theta, alpha, m, numIterations)  
  
print "theta: " ,theta 
```
