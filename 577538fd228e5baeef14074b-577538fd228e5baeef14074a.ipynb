疑问：预亏和预增对于股价的影响可能已经反应在公告出来之前。

可能的解释，

第一，行业性整体不行，很容易被投资者看到；

第二，利用公开信息和合法的调研已经足以发现公司经营行不行；

第三，内幕信息。

研究思路：

通过比较公告前5天和公告前14天的收盘价格，比较公告后5天和公告后14天的收盘价格。 附注：文中简单粗暴的timedelta 14会有很多缺陷，所以是研究是粗略的。

发现，都是50%左右的概率大于或者小于。

```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime
```

```python
basics = DataAPI.EquGet(equTypeCD=u"A",secID=u"",ticker=u"",listStatusCD=u"L,S",field=u"",pandas="1")
basics.head(3)
```

forecastType 11 预计亏损，21 预计盈利，22盈利增加，12 亏损减少，24 盈利下降，23 盈利减缓，13 下降减缓，31 业绩不确定，41 基本持平，51 经营预期及其他

```python
counter_before = 0
total = 0
counter_after = 0
for secID in basics['secID']:
    forecast = DataAPI.FdmtEfGet(ticker=u"",secID=secID,reportType=u"",endDate=u"",beginDate=u"",publishDateEnd=u"",publishDateBegin=u"",forecastType="22",field=u"",pandas="1")
    if not forecast.empty:
        for dt in forecast['publishDate']:
            beginDate = datetime.datetime.strptime(dt,'%Y-%m-%d %H:%M:%S') - datetime.timedelta(14)
            endDate = datetime.datetime.strptime(dt,'%Y-%m-%d %H:%M:%S')
            bd = beginDate.strftime('%Y%m%d') 
            ed = endDate.strftime('%Y%m%d')
            raw = DataAPI.MktEqudGet(tradeDate=u"",secID=secID,ticker=u"",beginDate=bd,endDate=ed,isOpen="",field=u"",pandas="1")
            if raw['closePrice'][-5:].mean() > raw['closePrice'].mean():
                counter_before += 1

            beginDate = datetime.datetime.strptime(dt,'%Y-%m-%d %H:%M:%S') 
            endDate = datetime.datetime.strptime(dt,'%Y-%m-%d %H:%M:%S') + datetime.timedelta(14)
            bd = beginDate.strftime('%Y%m%d') 
            ed = endDate.strftime('%Y%m%d')
            raw = DataAPI.MktEqudGet(tradeDate=u"",secID=secID,ticker=u"",beginDate=bd,endDate=ed,isOpen="",field=u"",pandas="1")
            if raw['closePrice'][:5].mean() < raw['closePrice'].mean():
                counter_after += 1
            total +=1

print(counter_before,counter_after,total)
```
