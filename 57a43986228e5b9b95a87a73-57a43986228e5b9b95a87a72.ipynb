1. ROE 系列
步骤：
1. 筛选出某月创历史新高的股票
2. 取该月及之后3-4个月的股票收盘价、当月之前每月最高收盘价——>得到当月之前历史最高价和当月及之后的收盘价
3. 取创新高股票对应的ROE值，按5%以下，5%-8%，8%-15%,15%以上分几个档次研究
4. 对于每个列表，取创历史新高的当天股价、之后至100天的股价，取二者比值，画出走势图

```python
import pandas as pd
import datetime
import numpy as np
from CAL.PyCAL import *
#因为优矿的ROE（TTM） 与Wind上的 ROE（TTM）的数据不一样，所以这里调整为先使用Wind上的数据分组后上传至优矿，进行后续验证
data=pd.read_excel('20140131hignest.xlsx')#这个是我在Wind上取得201401月份创历史新高的股票文件，含三列：股票代码、股票简称、ROE（TTM）；如果不做ROE的话，其他的因子可以值在优矿上取到
data['ticker']=data['code'].apply(lambda x: x[0:6])
data_new=data.drop(['code'],axis=1)
#获取股票自上市以来每月最高收盘价
pieces=[]
for stc in data_new.ticker.tolist() :
    #stc="u\'"+stc+"\'"
    datah=DataAPI.MktEqumAdjGet(secID=u"",ticker=stc.decode("utf-8"),monthEndDate=u"",isOpen=u"1",beginDate=u"",endDate=u"20131231",field=u"highestPrice",pandas="1")
    maxp=max(datah['highestPrice'])
    #print maxp,stc
    #当月起每个交易日收盘价
    data_main=DataAPI.MktEqudAdjGet(tradeDate=u"",secID=u"",ticker=stc.decode("utf-8"),isOpen=u"",beginDate=u"20140101",endDate=u"20140531",field=u"secID,tradeDate,closePrice",pandas="1")
    #print data_main#.T
    #把最高价合到data_main上
    a=data_main.iloc[0]
    a['closePrice']=maxp
    a['tradeDate']='higest_old'
    a=pd.DataFrame(a)
    data_main=data_main.append(a.T,ignore_index=True)
    #print len(data_main)#长度是一样的
    data_main=pd.DataFrame(data_main.closePrice.values,index=data_main.tradeDate,columns=[stc])
    pieces.append(data_main)
data_body=pd.concat(pieces,axis=1)
#data_body['mean']=data_body.mean(axis=1)#验证方式data_body.head(1).T.sum()
#data_body.columns.tolist()
#把data_new的ROE合到data_body上
data_new1=pd.DataFrame(data_new.ROE.values,index=data_new.ticker,columns=['ROE'])
data_new1=data_new1.T
data_last=(data_body.append(data_new1).T).dropna()#.T

#排序分组
data_last=data_last.sort(columns='ROE')

data_level1=(data_last[data_last['ROE']<5]).T
data_level1['mean_lv1']=data_level1.mean(axis=1)#pandas.core.series.Series
data_level1_1=data_level1['mean_lv1']
df1=pd.DataFrame(data_level1_1)

data_level2=(data_last[(data_last['ROE']>=5) & (data_last['ROE']<8)]).T
data_level2['mean_lv2']=data_level2.mean(axis=1)
data_level2_2=data_level2['mean_lv2']
df2=pd.DataFrame(data_level2_2)
df=pd.concat([df1,df2],axis=1)

data_level3=(data_last[(data_last['ROE']>=8) & (data_last['ROE']<15)]).T
data_level3['mean_lv3']=data_level3.mean(axis=1)
data_level3_3=data_level3['mean_lv3']
df3=pd.DataFrame(data_level3_3)
df=pd.concat([df,df3],axis=1)

data_level4=(data_last[data_last['ROE']>15]).T
data_level4['mean_lv4']=data_level4.mean(axis=1)
data_level4_4=data_level4['mean_lv4']
df4=pd.DataFrame(data_level4_4)
df=pd.concat([df,df4],axis=1)
df=df.drop(['ROE'])#((df.T).drop(['ROE'],axis=1)).T#我发神经病时候的做法
#把最后一行移到第一行
df=df.T
highest_old=df.pop('higest_old')
df.insert(0,'higest_old',highest_old)
df=df.T

#df['date']=df.index
num=len(df)
#取创历史新高月份股价第一次大于历史最高价
df=pd.DataFrame(df.values,index=range(num),columns=['mean_lv1','mean_lv2','mean_lv3','mean_lv4'])#,'date'#df=df.drop('date',axis=1)
#df
#取每列第一次大于highest_old的数据及其之后的数据
max_l=[]
max_l1=df.loc[df.index[0],'mean_lv1']
max_l.append(max_l1)
max_l2=df.loc[df.index[0],'mean_lv2']
max_l.append(max_l2)
max_l3=df.loc[df.index[0],'mean_lv3']
max_l.append(max_l3)
max_l4=df.loc[df.index[0],'mean_lv4']
max_l.append(max_l4)
#max_l
#df.iloc[:,0]#.tolist()[i]
#m=[]

df_s=[]
for i in range(len(df.columns)):    
    a=1
    b=(df.iloc[:,i]).tolist()
    while (b[a]< max_l[i]):
        a=a+1
        
    df_s.append(df.iloc[a:,i])
    #print df_s
    #m.append(a)
df_c=pd.concat(df_s,axis=1)
#df_c
df_c.iloc[:,1].dropna().tolist()
e=[]
for i in range(4):
    e.append(pd.Series(df_c.iloc[:,i].dropna().tolist(),index=range(len(df_c.iloc[:,i].dropna().tolist()))))
#e
df_f=pd.concat(e,axis=1)
df_f.columns=['mean_lv1','mean_lv2','mean_lv3','mean_lv4']
df_f=df_f.dropna()
#df_f
#算比值
num1=len(df_f)
ratio=[]
for i in range(num1):
    ratiobody=[]
    retnratio1=df_f['mean_lv1'][i]/df_f['mean_lv1'][0]
    ratiobody.append(retnratio1)
    
    retnratio2=df_f['mean_lv2'][i]/df_f['mean_lv2'][0]
    ratiobody.append(retnratio2)
    
    retnratio3=df_f['mean_lv3'][i]/df_f['mean_lv3'][0]
    ratiobody.append(retnratio3)
    
    retnratio4=df_f['mean_lv4'][i]/df_f['mean_lv4'][0]
    ratiobody.append(retnratio4)
    
    ratio.append(ratiobody)

df_new=pd.DataFrame(ratio,index=df_f.index,columns=['mean_lv1','mean_lv2','mean_lv3','mean_lv4'])#index=df.index#range(num)
#df_new=df_new.head(100)
#df_new.plot(figsize=(15,6))#用来画图的
df_new.to_excel('20140131_pass.xlsx')#导出成excel的
```
