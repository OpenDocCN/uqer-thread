代码如下：（代码可能写得不够简练）

```python
## 代码运行时间略长,15min左右。一次运行后存为csv
import pandas as pd

def CalcNextDate(date,date_list):            ## 得到下一个月日期
    for i in range(len(date_list)-1):
        if date==date_list[i]:
            return date_list[i+1]

universe = set_universe('HS300')
data=DataAPI.TradeCalGet(exchangeCD=u"XSHE",beginDate=u"20110801",endDate=u"20150801",field=['calendarDate','isMonthEnd'],pandas="1")   ## 得到日期列表
data = data[data['isMonthEnd'] == 1]
date_list_ext = data['calendarDate'].values.tolist()
date_list = date_list_ext[:-1]  ## 最后一个月无法计算下个月的收益 

df = pd.DataFrame()

## 获取所有因子数据，以便选取不同因子来聚类
for date in date_list:
    temp = DataAPI.MktStockFactorsOneDayGet(tradeDate=date,secID=universe,ticker=u"",field=u"",pandas="1")    
    df = pd.concat([df,temp])

    
df['CHGPct']=pd.Series('x',index=df.index)               ## 添加指数月变化率列
df['MonthGain']=pd.Series('x',index=df.index)           ## 添加月收益率列
length = len(df)
newindex=[]
for i in range(length):
    newindex.append(i)
df.index = newindex 

## 获取当月及下月股价、指数值，计算月收益率、指数变化率
for i in range(length):
    temp = df[i:i+1]
    temp_secID = df[i:i+1]['secID'].values[0]
    temp_Date = df[i:i+1]['tradeDate'].values[0]
    next_Date = CalcNextDate(temp_Date,date_list_ext)
    try:
        index1 = DataAPI.MktIdxdGet(tradeDate=temp_Date,indexID=u"",ticker=u"000300",beginDate=u"",endDate=u"",field=u"closeIndex",pandas="1")[0:1][['closeIndex']].values[0][0]
        index2 = DataAPI.MktIdxdGet(tradeDate=next_Date,indexID=u"",ticker=u"000300",beginDate=u"",endDate=u"",field=u"closeIndex",pandas="1")[0:1][['closeIndex']].values[0][0]
        price1 =DataAPI.MktEqudGet(tradeDate=temp_Date,secID=temp_secID,ticker=u"",beginDate=u"",endDate=u"",isOpen="",field=['closePrice'],pandas="1")[0:1][['closePrice']].values[0][0]
        price2 = DataAPI.MktEqudGet(tradeDate=next_Date,secID=temp_secID,ticker=u"",beginDate=u"",endDate=u"",isOpen="",field=['closePrice'],pandas="1")[0:1][['closePrice']].values[0][0]
        CHGPct = (index2-index1)/index1
        MonthGain = (price2-price1)/price1
    except:
        pass
    df['CHGPct'][i] = CHGPct
    df['MonthGain'][i] = MonthGain
df.to_csv('rawdata.csv')
```

## 二、聚类
关于聚类的方法，层次聚类和k-means比较常见，我在这里选用了最常见的k-means方法，具体关于k-means的介绍网上比较多，就不再介绍了。<br>
这里并没有使用sklearn的KMeans，而是以前做其他数据的用过的版本。代码如下：

```python
from __future__ import division
import numpy as np
import matplotlib.pyplot as plt

## 计算距离
def CalcDist(v1,v2):              
    return np.sqrt(np.sum(np.power(v1 - v2,2)))

## 初始化中心
def initCenter(data,k):
    numSamples,dim = data.shape
    print data.shape
    centers = np.zeros((k,dim))
    for i in range(k):
        index = int(np.random.uniform(0,numSamples))
        centers[i,:] = data[index,:]
    return centers

## 主要部分
def kmeans(data,k):
    numSamples = data.shape[0]
    clusterAssment = np.mat(np.zeros((numSamples,2)))
    clusterChanged = True
    centers = initCenter(data,k)
    
    while clusterChanged:
        clusterChanged = False
        for i in xrange(numSamples):  
            minDist  = 100000.0  
            minIndex = 0  
            
            for j in range(k):  
                distance = CalcDist(centers[j, :], data[i, :])  
                if distance < minDist:  
                    minDist  = distance  
                    minIndex = j  
               
            if clusterAssment[i, 0] != minIndex:  
                clusterChanged = True  
                clusterAssment[i, :] = minIndex, minDist**2  
    
        for j in range(k):  
            pointsInCluster = data[np.nonzero(clusterAssment[:, 0].A == j)[0]]  
            centers[j, :] = np.mean(pointsInCluster, axis = 0)  
  
    print 'Congratulations, cluster complete!'  
    return centers, clusterAssment  

## 可视化
def showCluster(dataSet, k, centroids, clusterAssment):  
    numSamples, dim = dataSet.shape  
    if (dim != 2) and (dim!=3):  
        print "Sorry! I can not draw because the dimension of your data is not 2!"  
        return 1  
  
    mark = ['or', 'ob', 'og', 'ok', '^r', '+r', 'sr', 'dr', '<r', 'pr']  
    if k > len(mark):  
        return 1  
    if dim==2:
    ##  画散点
        for i in xrange(numSamples):  
            markIndex = int(clusterAssment[i, 0])
            plt.plot(dataSet[i, 0], dataSet[i, 1], mark[markIndex])  
      
        mark = ['Dr', 'Db', 'Dg', 'Dk', '^b', '+b', 'sb', 'db', '<b', 'pb']  
        # 标记中心  
        for i in range(k):
            plt.plot(centroids[i, 0], centroids[i, 1], mark[i], markersize = 12)  
      
        plt.show()
```

上述代码可视化只支持2维，就先试试两个因子吧，选取PE、LFLO进行聚类

```python
PE = np.array(df['PE'])                         ##市盈率
LFLO = np.array(df['LFLO'])                 

CHG = np.array(df['CHGPct'])
MG = np.array(df['MonthGain'])

##  构建输入矩阵
used_data = [PE,LFLO]
data = np.mat(used_data)
data = data.T

k =6    ## 类型数
## 开始聚类
centers,clusterAssment = kmeans(data,k)
showCluster(data,k,centers,clusterAssment)
print centers
```
