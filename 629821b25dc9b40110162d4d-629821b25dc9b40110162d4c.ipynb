# 月频测试

```python
"""
月频策略---想法与统计判断，细节分析与进一步挖掘提升。
基本面财务遗传算法 财务数据-- 分析问题为 每一个行业内，有驱动因素。
1. 一个行业包含多个细胞，具有一个基因组
2. 每个基因组由3分DNA组成（1.极值剔除（数据、上端比例、下端比例） 
                         2.分组判断（数据））
                         3.单调（数据、正负项判断）
3. 每个cross section 获取所有基金组（每个基金组有效性、残差序列）
4. 首先当前初始迭代包含上期最强n个基因组（10个）*这个基金具有3倍传染性，上期其他行业最强1个基因组（30个），新基因组（20个）
4.1 计算初始代解释程度，剔除
4.2 杂交（剔除秩相关性数不足一个阈值的）* n （直到一个条件）
4.3 根据一个标准优选（解释1期delta(去年同比)变化，根据delta变化与估值变化，相关性作为大小，并保存为权重）
5.买入每个行业deep value（购买整体 加权平均截面解释弱的行业，比如平均相关分之一 或者 cos[pi/2]）
"""
import pandas as pd
import numpy as np
import datetime
import random
from scipy.stats import spearmanr
#固定基本证券研究池子,同时确定其行业属性
try:
    data = pd.read_csv('stk_pe/stk_list.csv',index_col=0)
except:
    q=DataAPI.EquGet(equTypeCD=u"A",listStatusCD=u"",exchangeCD="",ListSectorCD=u""
                     ,field=u"secID,ticker,ListSector,secShortName",pandas="1")
    #仅包含主板
    q = q.loc[q.iloc[:,2] == q.iloc[0,2]]
    q.index = q['secID']
    q =  q.drop('secID', 1)
    #获取行业
    w  = DataAPI.EquIndustryGet(secID=q.index,industryVersionCD=u"010321",intoDate=
                                '20220501',field=u"secID,industryID1",pandas="1")
    w.index = w['secID'] 
    q = q.loc[set(w['secID']).intersection(q.index)]
    q['industry'] = w.loc[q.index,w.columns[1]]
    q.to_csv('stk_pe/stk_list.csv')
    data = q.copy()
    del q,w
```

```python
# 直接使用财务指标，分析财务指标解释能力
def season_adj(year,n):
    if n==1:
        return([str(year)+'0101',str(year)+'0401'])
    if n==2:
        return([str(year)+'0401',str(year)+'0701'])
    if n==3:
        return([str(year)+'0701',str(year)+'1001'])
    if n==4:
        return([str(year)+'1001',str(year+1)+'0101'])
def ps_data(stk_code, year, season):
    #
    [bd,ed] = season_adj(year,season)
    #成长能力、盈利能力、营运能力
    all_d = pd.DataFrame(index=stk_code)
    data = DataAPI.FdmtIndiGrowthPitGet(secID=stk_code,endDate=ed,beginDate=bd,field=u"",pandas="1")
    data.index = data['secID']
    data = data.drop(data.columns[[1,2,3,4,6,8]],1)
    all_d = pd.merge(all_d,data,left_index=True,right_index=True)
    #盈利能力
    data = DataAPI.FdmtIndiRtnPitGet(secID=stk_code,endDate=ed,beginDate=bd,field=u"",pandas="1")
    data.index = data['secID']
    data = data.drop(data.columns[[0,1,2,3,4,5,6,8]],1)
    all_d = pd.merge(all_d,data,left_index=True,right_index=True)
    #运营能力
    data = DataAPI.FdmtIndiTrnovrPitGet(secID=stk_code,endDate=ed,beginDate=bd,field=u"",pandas="1")
    data.index = data['secID']
    data = data.drop(data.columns[[0,1,2,3,4,5,6,8]],1)
    all_d = pd.merge(all_d,data,left_index=True,right_index=True)
    all_d = all_d.drop_duplicates()
    all_d['publishDate'] = pd.to_datetime(all_d['publishDate'])
    all_d = all_d.sort_values('publishDate', ascending=False)
    all_d.index.name = str(year)+'Q'+str(season)
    return(all_d)
def data_fresh(d1,d2,now_t,s):
        try:
            d1 = pd.read_csv('section_data/'+str(now_t.year)+'Q'+str(s)+'.csv', index_col =0)
        except:
            d1 = ps_data(data.index, now_t.year, s)
            d1.to_csv('section_data/'+str(now_t.year)+'Q'+str(s)+'.csv')
        try:
            d2 = pd.read_csv('section_data/'+str(now_t.year-1)+'Q'+str(s)+'.csv', index_col =0)
        except:
            d2 = ps_data(data.index, now_t.year-1, s)
            d2.to_csv('section_data/'+str(now_t.year-1)+'Q'+str(s)+'.csv')
        return()
def fresh(d1,d2,now_t):
    if now_t < datetime.datetime(now_t.year, 5, 1):
        #使用上年3季度报
        if d1.index.name == str(now_t.year-1)+'Q3' and d2.index.name == str(now_t.year-2)+'Q3':
            return(d1,d2)
        else:
            data_fresh(d1,d2,now_t,3)
    else:
        if now_t <datetime.datetime(now_t.year, 9, 1):
            if d1.index.name == str(now_t.year)+'Q1' and d2.index.name == str(now_t.year-1)+'Q1':
                return(d1,d2)
            else:
                data_fresh(d1,d2,now_t,1)
        else:
            if d1.index.name == str(now_t.year)+'Q2' and d2.index.name == str(now_t.year-1)+'Q2':
                return(d1,d2)
            else:
                data_fresh(d1,d2,now_t,2)
    return(d1,d2)
    
```

```python
def xx():
    return()
def random_create(n):
    res = []
    for i in range(n):
        n3 = random.sample(range(38),3)
        if random.randint(0,9)<=6:
            sks = 2
        else:
            if random.randint(0,2)<=1:
                sks =3
            else:
                sks =4
        if random.randint(0,9)<=5:
            skt = 0
        else:
            if random.randint(0,3)<=2:
                skt =0.05
            else:
                skt =0.1
        if random.randint(0,9)<=5:
            ska = 0
        else:
            if random.randint(0,3)<=2:
                ska =0.05
            else:
                ska =0.1
        res.append([n3[0],ska,skt,n3[1],sks,n3[2]]+[np.NAN]*9)
    return(res)
def cross_test(df, delta, mat):
    mat.index = range(len(mat))
    for i in range(len(mat)):
        if mat.loc[mat.index[i],'already'] == 1:
            continue
        else:
            mat.loc[mat.index[i],'already'] = 1
            mat.iloc[i,6:13] == np.NAN
        tdf=df.copy()
        #剔除 若小于最小单元1，则不剔除
        if pd.isnull(mat.iloc[i,1]) or mat.iloc[i,1]*len(tdf)<1:
            xx()
        else:
            pop_n =int(mat.iloc[i,0])
            tdf = tdf.loc[tdf.iloc[:,pop_n]<=(tdf.iloc[:,pop_n].quantile(1-mat.iloc[i,1]))]
        if pd.isnull(mat.iloc[i,2]) or mat.iloc[i,2]*len(tdf)<1:
            xx()
        else:
            pop_n =int(mat.iloc[i,0])
            tdf = tdf.loc[tdf.iloc[:,pop_n]>=(tdf.iloc[:,pop_n].quantile(mat.iloc[i,2]))]
        mat.iloc[i,12] = len(tdf)
        #分组 并计算排序
        fz_n = int(mat.iloc[i,3])
        pct = 1/mat.iloc[i,4]
        for k in range(int(mat.iloc[i,4])):
            kdf = tdf.loc[tdf.iloc[:,fz_n]<=tdf.iloc[:,fz_n].quantile(pct*(k+1))]
            kdf = kdf.loc[kdf.iloc[:,fz_n]>tdf.iloc[:,fz_n].quantile(pct*(k))]
            try:
                mat.loc[mat.index[i],'avg_abs_r'+str(k+1)] = spearmanr(kdf.iloc[:,int(mat.iloc[i,5])],
                          kdf.iloc[:,-1],nan_policy='omit').correlation
            except:
                mat.loc[mat.index[i],'avg_abs_r'+str(k+1)] = np.NAN
        #计算 delta情况
        t_delta = delta.loc[tdf.index].copy()
        try:
            mat.loc[mat.index[i],'delta_w'] = spearmanr(t_delta.iloc[:,int(mat.iloc[i,5])],
                                                    t_delta.iloc[:,-1],nan_policy='omit').correlation
        except:
            mat.loc[mat.index[i],'delta_w'] = np.NAN
    mat['avg_abs_r'] = mat.iloc[:,6:10].abs().mean(1)
    return(mat)
def shuffle_gene(gene):
    s_code =range(10)
    random.shuffle(s_code)
    new_gene = pd.DataFrame(index=list(range(10,20)),columns=gene.columns)
    new_gene.iloc[:,0:3] = gene.iloc[s_code,0:3].values
    random.shuffle(s_code)
    new_gene.iloc[:,3:5] = gene.iloc[s_code,3:5].values
    random.shuffle(s_code)
    new_gene.iloc[:,5] = gene.iloc[s_code,5].tolist()
    new_gene['already'] = [0]*10
    return(gene.append(new_gene))
```

```python
ind_kind = list(set(data['industry']))
gene_m = pd.read_csv('origin_gene.csv',index_col=0)
for t_ind in range(len(ind_kind)):
    t_ind_s = data.index[data['industry']==ind_kind[t_ind]]
    t_ind_delta = delta.loc[set(t_ind_s).intersection(delta.index)]
    t_ind_df = now_c.loc[t_ind_delta.index]
    mat = gene_m.loc[gene_m['ind_n']==ind_kind[t_ind]].copy()
    #随机添加10个 & 新生成几个新家伙
    mat = mat.append(gene_m.iloc[random.sample(range(len(gene_m)),10)])
    mat = mat.append(pd.DataFrame(random_create(10),columns=mat.columns))
    mat['already'] = [0]*len(mat)
    mat['ind_n'] = [np.NAN]*len(mat)
    """
    检测验证
    """
    #迭代100轮（剔除平均系数最低20只，然后杂交生成10只，突变10只）
    for generation in range(100):
        mat = cross_test(t_ind_df, t_ind_delta, mat)
        mat = mat.drop_duplicates()
        mat = mat.sort_values('avg_abs_r',ascending=False).iloc[:10]
        mat.index= range(10)
        mat = shuffle_gene(mat)
        mat = mat.append(pd.DataFrame(random_create(10),columns=mat.columns))
    mat =cross_test(t_ind_df, t_ind_delta, mat)
    mat = mat.drop_duplicates()
    mat = mat.sort_values('avg_abs_r',ascending=False).iloc[:10]
    mat['ind_n'] = [ind_kind[t_ind]] * len(mat)
    try:
        all_gene = all_gene.append(mat)
    except:
        all_gene = mat
all_gene.to_csv('gene/'+now_date.strftime('%Y%m%d')+'.csv')
```
