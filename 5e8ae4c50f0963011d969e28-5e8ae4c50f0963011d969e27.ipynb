# 20200406-MarkDown的使用方法

# 第一部分：标题级别


## 第一种方法，\#字符


>1. 用#字符开头来表示一个标题字符大小的级别，一共6级
>2. 可以在标题的尾部追加同等数量的#来让格式显得更加整齐

### 范例1.1

\# 一级标题
\## 二级标题
\### 三级标题
\#### 四级标题
\##### 五级标题
\###### 六级标题
无标题

># 一级标题
>## 二级标题
>### 三级标题
>#### 四级标题
>##### 五级标题
>###### 六级标题
>无标题

### 范例1.2

\# 一级标题 \#
\## 二级标题 \##
\### 三级标题 \###
\#### 四级标题 \####
\##### 五级标题 \#####
\###### 六级标题 \###### 
无标题

># 一级标题 #
>## 二级标题 ##
>### 三级标题 ###
>#### 四级标题 ####
>##### 五级标题 #####
>###### 六级标题 ######
>无标题

## 第二种方法：=字符或-字符

>在字符下一行写若干=字符或-字符来表示一个标题字符大小的级别，一共2级

### 范例1.3
一级标题
\===
二级标题
\----
无标题

> 一级标题
> ====
> 二级标题
> ---
> 无标题


# 第二部分：正文文本

## 2.1 文本块
> 用>字符把某块内容放在一起

### 范例2.1
\> 我毕业于中欧商学院
\> 我爱中欧商学院

> 我毕业于中欧商学院
> 我爱中欧商学院

### 范例2.2
\>> 我毕业于中欧商学院
\>> 我爱中欧商学院

>> 我毕业于中欧商学院
>> 我爱中欧商学院

## 2.2 斜体或者加粗

> 斜体：在待倾斜的文字的两边加上一个*(星号)或者_(下划线)
> 加粗：在待加粗的文字的两边加上两个*(星号)或者_(下划线)


### 范例2.3
\> 我毕业于\*中欧\*商学院
\> 我爱\*\*中欧\*\*商学院

> 我毕业于*中欧*商学院
> 我爱**中欧**商学院

### 范例2.4
\> 我毕业于\_中欧\_商学院
\> 我爱\_\_中欧\_\_商学院

> 我毕业于_中欧_商学院
> 我爱__中欧__商学院

不同的markdown编辑器对markdown语法的支持程度不尽相同，比如这里的_(下划线)不起作用

## 2.3 无序列表
> 介绍某个事物时，需要分条去介绍
> 在*(星号)、+(加号)或-(减号)特殊字符后面跟一个空格

### 范例2.5
\* 我毕业于中欧商学院
\* 我爱中欧商学院
\* 中欧商学院让我眼界更加开阔

> * 我毕业于中欧商学院
> * 我爱中欧商学院
> * 中欧商学院让我眼界更加开阔

### 范例2.6
\+ 我毕业于中欧商学院
\+ 我爱中欧商学院
\+ 中欧商学院让我眼界更加开阔

> + 我毕业于中欧商学院
> + 我爱中欧商学院
> + 中欧商学院让我眼界更加开阔

## 2.4 有序列表
> 相对于无序列表，有序列表更关注先后顺序
> 在列表项的前面使用一个数字跟一个点，再加一个空格
> 数字的具体值并不起作用，它仅标识这是一个有序列表。

### 范例2.7
1.我毕业于中欧商学院
1.中欧商学院让我眼界更加开阔，学到很多知识
1.我爱中欧商学院

> 1. 我毕业于中欧商学院
> 1. 中欧商学院让我眼界更加开阔，学到很多知识
> 1. 我爱中欧商学院

### 推荐方法

> 在上面的范例中，虽然左边的编号都是1，但是右边的编号是依次递增的。
> 如果也需要左边的序号是递增的，可以修改为如下格式：

### 范例2.8
1.很多同学都毕业于中欧商学院
2.大家在学校学到了很多本领
3.我们都爱中欧商学院

> 1. 很多同学都毕业于中欧商学院
> 2. 大家在学校学到了很多本领
> 3. 我们都爱中欧商学院

## 2.5 超链接
> 有时候需要引用一些网址信息，比如参考文档的连接，官网地址等等，可以使用url地址进行备注。
> 使用尖括号对把地址包围起来，格式:<url 地址>
> 根据markdown 的解析规则，它会被解析成为这样:<a href="url 地址">url 地址</a>

### 范例2.9

中欧商学院
<https://cn.ceibs.edu/>

### 推荐方法

> 上面格式不是很友好，我们可以使用这种格式:\[网站名](网站地址)

### 范例2.10

我爱\[中欧商学院](https://cn.ceibs.edu/)

> 我爱[中欧商学院](https://cn.ceibs.edu/)

## 2.6 图片

> 有些时候需要引入图片，在markdown中添加图片的方法是:!\[图片名称](图片地址)
> 或者拖动图片到markdown文本框中

## 2.7 转义

> 有些符号在markdown中有特殊的含义，比如#和*，想使用这些符号时，该怎么办?
> 可以使用转义符号反斜线(\\)对它们进行转义，也就是用\\#表示#，用\\\*表示*。

### 范例2.11

不使用转义
\# 你好
> # 你好

使用转义
\\# 你好
> \\# 你好

## 3.2 代码块

> 我们时常需要在文档中引入一段代码，可以使用三个命令号开头，三个命令号结尾，在中间书写代码内容。
> 命令号是\`，在Esc键和Tab键之间。
    
### 范例3.2
    
\```
代码1：我毕业于中欧商学院
代码2：中欧商学院让我眼界更加开阔，学到很多知识
代码3：我爱中欧商学院
\```


```
代码1：我毕业于中欧商学院
代码2：中欧商学院让我眼界更加开阔，学到很多知识
代码3：我爱中欧商学院
```
    
## 3.3 兼容html

markdown文件非常适合通过互联网展示给其他人。
markdown可以有效地兼容html，我们可以在markdown中使用html语法。

### 范例3.3
输入span标签，并设置红色字体和30号字体大小
\< span style="color:red;font-size:30px"\>
强烈推荐中欧商学院
\</span>
    
> <span style="color:red;font-size:30px">
> 强烈推荐中欧商学院
> </span>
    
    
不同的markdown编辑器对markdown语法的支持程度不尽相同，比如这里就没有显示红色和30号字体大小
