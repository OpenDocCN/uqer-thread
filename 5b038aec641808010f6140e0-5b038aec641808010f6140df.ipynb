```python
###Data   Loading stage
####################################################################################
import pandas as pd
###stock pool----hs300
hs300=DataAPI.IdxConsGet(secID=u"",ticker=u"000300",isNew=u"",intoDate=u"",field=u"",pandas="1")
hs300_new=hs300[hs300.isNew==1]
tickers_pool=list(hs300.consTickerSymbol.values)
##get monthly return for all stock
month_return=DataAPI.MktEqumGet(secID=u"",ticker=tickers_pool,monthEndDate=u"",beginDate=u"20160101",endDate=u"20170630",field=['secID','endDate','return'],pandas="1")
month_return['endDate']=month_return['endDate'].apply(lambda x:x.replace("-",""))


##obtain daily candidate_factor data for 2015-01 to 2017-06,check their missing value, delete those factors who has large missing vlue
candidate_factor=['ROE','DAVOL10','DHILO','EBITToTOR','PB','PE','GrossIncomeRatio','ASSI','BLEV','DebtEquityRatio','ETOP','DAVOL5']

new_factor=['ticker','tradeDate']+candidate_factor
factor_list=[]
for each_ticker in tickers_pool:
    factor_frame=DataAPI.MktStockFactorsDateRangeGet(secID=u"",ticker=each_ticker,beginDate=u"20160101",endDate=u"20170630",field=new_factor,pandas="1")
    factor_frame['tradeDate']=factor_frame['tradeDate'].apply(lambda x:x.replace("-",""))
    factor_frame['secID']=factor_frame['ticker'].apply(lambda x:x+".XSHG")
    factor_list.append(factor_frame)
            
factor_df=pd.concat(factor_list)  
##check their missing values
#print(factor_df.isnull().sum())

##############################################################################################

##Calculate factor mean for each (industry,tradedate)

###equity-industry map
equity_industry=DataAPI.EquIndustryGet(industryVersionCD=u'010303',industry=u"",secID="",ticker=tickers_pool,intoDate=u"",field=[u'ticker','secShortName','industry','intoDate','outDate','industryName1','isNew'],pandas="1")
equity_ind=equity_industry[equity_industry['isNew']==1]
equity_ind['secID']=equity_ind['ticker'].apply(lambda x:x+".XSHG")
equity_ind=equity_ind[['secID','industryName1']]
#print(equity_ind)

####calculate industry_mean    
factor_ind=pd.merge(factor_df,equity_ind,on='secID',how='left')
factor=factor_ind.drop('ticker',axis=1,inplace=True)
#print(factor_ind)
factor_new_name=[name+'_mean'for name in candidate_factor]
mean_indus=factor_ind.groupby(['industryName1','tradeDate'])[candidate_factor].mean().reset_index()

######merge two dataset so that for each (stock,tradedate) their industry average factor value is also known 
old_new_name={}
for i in range(len(candidate_factor)):
    old_new_name[candidate_factor[i]]=factor_new_name[i]

mean_indus.rename(columns=old_new_name,inplace=True)    
#print(factor_ind)
factor_frame=pd.merge(factor_ind,mean_indus,on=['industryName1','tradeDate'],how='left')
#print(factor_frame)
for fac in candidate_factor:
    factor_frame[fac].fillna(factor_frame[old_new_name[fac]])
            
factor_frame.dropna(inplace=True) 
#print(factor_frame)            
###Obtain calendar inform,we want to the information of end of each month and their previous end_date.
####the reason is we want to build a model for  factors of previous_month_end say ,March 31 to predict the monthly return of next_month April
###However, Uqer returns unicode format,we want dataframe
#---------------------------------------------------------------------------------------------------
#calendar=DataAPI.TradeCalGet(exchangeCD=u"XSHG,XSHE",beginDate=u"20160101",endDate=u"20170601",field=u"",pandas=1)
#calendar=calendar[calendar.isMonthEnd==1][['calendarDate','isMonthEnd']]
#calendar['calendar_date']=calendar['calendar_date'].apply(lambda x:x.replace("-",""))
#calendar['previous_month_end']=calendar['calendar_date'].shift(1)
#calendar=calendar[['calendarDate','previous_month_end']]
#calendar.rename(columns={"calendarDate":"month_end"},inplace=True)

################hardcode calendar dataframe
month_end=['20150131',
'20150228',
'20150331',
'20150430',
'20150531',
'20150630',
'20150731',
'20150831',
'20150930',
'20151031',
'20151130',
'20151231',
'20160131',
'20160229',
'20160331',
'20160430',
'20160531',
'20160630',
'20160731',
'20160831',
'20160930',
'20161031',
'20161130',
'20161231',
'20170131',
'20170228',
'20170331',
'20170430',
'20170531',
'20170630',
'20170731',
'20170831',
'20170930',
'20171031',
'20171130',
'20171231',
'20180131',
'20180228',
'20180331',
'20180430']
previous_month_end=[np.nan,
'20150131',
'20150228',
'20150331',
'20150430',
'20150531',
'20150630',
'20150731',
'20150831',
'20150930',
'20151031',
'20151130',
'20151231',
'20160131',
'20160229',
'20160331',
'20160430',
'20160531',
'20160630',
'20160731',
'20160831',
'20160930',
'20161031',
'20161130',
'20161231',
'20170131',
'20170228',
'20170331',
'20170430',
'20170531',
'20170630',
'20170731',
'20170831',
'20170930',
'20171031',
'20171130',
'20171231',
'20180131',
'20180228',
'20180331']
calendar=pd.DataFrame([month_end,previous_month_end]).T
calendar.rename(columns={0:'month_end',1:'previous_month_end'},inplace=True)
########Only retain the month_end information with 
####previous_Month's  factor==Next_Month return
m_factor=pd.merge(factor_frame,calendar,left_on=['tradeDate'],right_on=['previous_month_end'],how='right')
m_factor.dropna(inplace=True)

tfactor=pd.merge(m_factor,month_return,left_on=['secID','tradeDate'],right_on=['secID','endDate'],how='left')
tfactor.dropna(inplace=True)
#print(m_factor)
factor_name=candidate_factor


#################################################################################################################################
#Modeling and factor analysis Stage
#######Univariate analysis.
###########Calculate the IC for each factor, i.e the correlation of previous's month factor and return of next month
import scipy.stats as st
def cal_IC(df,factor):
    df_temp=df.copy() 
    ic=df_temp[factor].corr(df_temp['return'])
    return ic

factor_ic={}
for each_factor in factor_name:
    factor_ic[each_factor]=cal_IC(tfactor,each_factor)          
#print(factor_ic)
#################cal IC for single stock
####select factor with |IC|>=0.1
useful_factor=['ROE','PB','DAVOL5','DAVOL10']

###############################################################################################################################################################
##The next step is to build a  'Score' which combines multi-factor into a single measure ,once this 'Score' is obtained we can pick the top 1/3 stocks or last 1/3 stocks
###to compare their performance,Hopefully, with the right model tool, the resulted segementation will have a monotonic property, that is ,the higher the segment the better the 
###performance, if not, we shall at least expect the top segment outperform the buttom segment by a lot.



#####Here we use OLS model. The (x,y) are monthly factors for each stock and monthly return for that stock  ,training time period from 2015-01-01 to 2017-05-30
###the 'Score' is simply the predicted return from OLS
x=tfactor[useful_factor]
y=tfactor['return']
xx=x.as_matrix()
yy=y.as_matrix()


from sklearn.linear_model import LinearRegression

las = LinearRegression()
regr = las.fit(xx, yy)  

# print(regr.score(xx,yy))

coef_dict={}
for i in range(len(useful_factor)):
    coef_dict[useful_factor[i]]=regr.coef_[i]
    
#print(coef_dict)    

yy_score=regr.predict(xx)
y['pred']=yy_score
tfactor['predicted']=yy_score
###Set score to predicted_return
tfactor['score']=tfactor['predicted']

print(tfactor)

############################################################################################################################################################
######Backtesting stage
##按此分数分3组，每月出调仓，算出3组在2015-01-01到2017-05-30的表现
#### Data configuration
   
import time
from CAL.PyCAL import *
start_time=time.time()

start='2016-01-01'
end='2017-12-30'
benchmark='HS300'
universe=DynamicUniverse('HS300')
capital_base=10000000
fred='d'
refresh_rate=Monthly(1)
account = {'fantacy_account':AccountConfig(account_type='security',capital_base=capital_base)}

factor_data=tfactor[['secID','tradeDate','score']]
factor_data.rename(columns={'score':'factor'},inplace=True)
factor_data=factor_data.set_index('tradeDate',drop=False)
#factor_data.head(20)
factor_data.dropna(inplace=True)
q_dates=factor_data.index.values
q_dates=list(set(q_dates))
print(q_dates)

###-------------BACKTESTING-------------------------------------------------------------------------------------#
####在每月底把股票按此打分分三组，对第i个segment每次等仓位持有第i组的股票  ，我们希望这几组Portfolio表现递增。不选四个的原因是1，4组颜色很像 ^_^
sim_params=quartz.SimulationParameters(start,end,benchmark,universe,capital_base,refresh_rate=refresh_rate,accounts=account)
data=quartz.get_backtest_data(sim_params)
#运行结果
results={}
result_new={}
N=3
###遍历回测
for piece in range(1,N+1):
    def initialize(context):
        pass
    
    def handle_data(context):
        account=context.get_account('fantacy_account')
        current_universe=context.get_universe('stock',exclude_halt=True)
        pre_date=context.previous_date.strftime("%Y%m%d")
      
        if pre_date not in q_dates:
            return
                                           
        q=factor_data.ix[pre_date].dropna()

        q=q.set_index('secID',drop=False)
        q=q.ix[current_universe]
        #print(q)
        q.dropna(inplace=True)
        #print(q)
        
        #df.quantile deal with missing value automatically
        q_min=q['factor'].quantile((piece-1)*1.0/N)
        q_max=q['factor'].quantile(piece*1.0/N) 
   
        my_univ=q[(q['factor']>=q_min)&(q['factor']<q_max)].index.values   
       
        
    #trading part
        positions=account.get_positions()
        
        #buy
        buy_list=list(set(my_univ).difference(positions))
        print(pre_date)
        print(['buy']+buy_list)
        for stk in buy_list:
            account.order_pct_to(stk,1.0/max(len(buy_list),1))
         #sell   
        sell_list=list(set(positions).difference(my_univ))
        print(['sell']+sell_list)
        for stk in sell_list:
            account.order_to(stk,0)  
                                               
    
    ##生成策略对象                                            
    strategy=quartz.TradingStrategy(initialize,handle_data)                                         
    #开始回测   
    bt,perf=quartz.quick_backtest(sim_params,strategy,data=data)
                                                
    ####保存运行结果
    results[piece]={'max_drawdown':perf['max_drawdown'],'sharpe':perf['sharpe'],'alpha':perf['alpha'],'annualized_return':perf['annualized_return'],'bt':bt}   
                                         
###############################画图展示结果    

import seaborn as sns

def draw_graphics(results):    
    sns.set_style('white')

    fig = plt.figure(figsize=(10,8))
    fig.set_tight_layout(True)
    ax1 = fig.add_subplot(211)
    ax2 = fig.add_subplot(212)
    ax1.grid()
    ax2.grid()

    for qt in results:
        bt = results[qt]['bt']

        data = bt[[u'tradeDate',u'portfolio_value',u'benchmark_return']]
        data['portfolio_return'] = data.portfolio_value/data.portfolio_value.shift(1) - 1.0   # 总头寸每日回报率
        data['portfolio_return'].ix[0] = data['portfolio_value'].ix[0]/	capital_base - 1.0
        data['excess_return'] = data.portfolio_return - data.benchmark_return                 # 总头寸每日超额回报率
        data['excess'] = data.excess_return + 1.0
        data['excess'] = data.excess.cumprod()                # 总头寸对冲指数后的净值序列
        data['portfolio'] = data.portfolio_return + 1.0     
        data['portfolio'] = data.portfolio.cumprod()          # 总头寸不对冲时的净值序列
        data['benchmark'] = data.benchmark_return + 1.0
        data['benchmark'] = data.benchmark.cumprod()          # benchmark的净值序列
        results[qt]['hedged_max_drawdown'] = max([1 - v/max(1, max(data['excess'][:i+1])) for i,v in enumerate(data['excess'])])  # 对冲后净值最大回撤
        results[qt]['hedged_volatility'] = np.std(data['excess_return'])*np.sqrt(252)
        results[qt]['hedged_annualized_return'] = (data['excess'].values[-1])**(252.0/len(data['excess'])) - 1.0
        ax1.plot(data['tradeDate'], data[['portfolio']], label=str(qt))
        ax2.plot(data['tradeDate'], data[['excess']], label=str(qt))


    ax1.legend(loc=0)
    ax2.legend(loc=0)
    ax1.set_ylabel(u"净值", fontproperties=font, fontsize=16)
    ax2.set_ylabel(u"对冲净值", fontproperties=font, fontsize=16)
    ax1.set_title(u"因子不同分位数分组选股净值走势", fontproperties=font, fontsize=16)
    ax2.set_title(u"因子不同分位数分组选股对冲中证500指数后净值走势", fontproperties=font, fontsize=16)

    # results 转换为 DataFrame
    results_pd = pd.DataFrame(results).T.sort_index()

    results_pd = results_pd[[u'alpha', u'annualized_return', u'max_drawdown',u'sharpe',  
                             u'hedged_annualized_return', u'hedged_max_drawdown', u'hedged_volatility']]

    cols = [(u'风险指标', u'Alpha'), (u'纯股票多头时', u'年化收益'),
            (u'纯股票多头时', u'最大回撤'), (u'纯股票多头',u'shape_ratio'),(u'对冲后', u'年化收益'), (u'对冲后', u'最大回撤'), (u'对冲后', u'收益波动率')]
    results_pd.columns = pd.MultiIndex.from_tuples(cols)
    results_pd.index.name = u'三分位组别'
    return results_pd


draw_graphics(results)

######################################################################
####结果分析：通过该分数得到的分组最好那组表现不错，但没有呈现单调性





```
