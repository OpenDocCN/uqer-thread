```python
    #先卖出
    for stk in account.valid_secpos:
        if stk not in buylist:
            order_to(stk, 0)

    # 再买入
    total_money = account.referencePortfolioValue
    prices = account.referencePrice  
    for stk in buylist:
        if np.isnan(prices[stk]) or prices[stk] == 0:  # 停牌或是还没有上市等原因不能交易
            continue
        order_to(stk, int(total_money * 0.95/len(buylist) / prices[stk] /100)*100)
```

## `  2、前复权行情本身就可能存在未来函数`
### 使用前复权价格，不论回测开始时间、结束时间是何时，使用的数据都是基于今天（回测当天）或某个时间的复权因子进行前复权获得的价格，因此使用前复权价格进行回测，回测结果肯定有问题。
> 举个栗子
  比如要买入价格最低的10只股票，如果用了前复权价格，那么取到的可能不是当时价格最低的，而是因为这几只股票进行了拆分或者分红。这样就存在未卜先知的问题，你是提前知道这个股票要分红的。
  
>   再举个栗子，在相似公司之间互相搬砖的策略，如果两个相似公司你利用前复权价格回测，发现规律，然后在两个股票之间搬砖，那么可能也是错的。因为这两只股票在当时的价格可能并不是你取到的数据。（最好取未复权的数据验证一下）

## ` 3、10万和100万回测出来结果会不一样`

造成这个结果的原因是，你的买入股票太多，而价格高的股票占用了太多的资金，有些股票满足不了100股的买入限制，买入失败。

## ` 4、调仓周期大的回测，不同的起始日期结果可能完全不同`
如果一个策略的调仓周期是5天，那么你3月1日测试，和3月2日测试，可能结果是不一样的，而且很可能相差很大

如果你的交易逻辑是符合规律的，那么你不妨将策略放在不同的测试日期多试几次，如果交易思路本身是合乎逻辑的，那么多试几次的收益不会相差太多。否则可能只是刚好过拟合了。
