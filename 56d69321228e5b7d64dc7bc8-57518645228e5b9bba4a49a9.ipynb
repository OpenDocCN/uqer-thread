模式匹配的方法预测市场走势

想法就是用最近20天的走势来找历史上类似的走势，然后来预测接下来5天的走势。

由成交金额和收盘价来匹配，算出距离，距离最小为最相似。

```python
from datetime import datetime
#获取时间日期
now=datetime.now()
now=now.strftime("%Y%m%d")
print 'Today:%s ' %now
```

```python
import numpy as np
import pandas as pd
#dwt距离测算
def dtw(t,r):
    n=len(t)
    m=len(r)
    #帧匹配距离矩阵
    d=np.zeros((n,m))
    for i in range(n):
        for j in range(m):
            d[i][j]=(t[i]-r[j])*(t[i]-r[j])
    #累积距离矩阵
    realmax=10000000
    D=np.ones((n,m))*realmax
    D[0][0]=d[0][0]
    #动态规划
    for i in range(1,n):
        for j in range(0,m):
            D1=D[i-1][j]
            if j>0:
                D2=D[i-1][j-1]
            else:
                D2=realmax
            if j>1:
                D3=D[i-1][j-2]
            else:
                D3=realmax
            D[i][j]=d[i][j]+min(D1,D2,D3)
    return D[n-1][m-1]

#数据标准化
def normalized(a):
    m=np.mean(a)
    s=np.std(a)
    return (a-m)/s
#数据载入
data=DataAPI.MktIdxdGet(ticker='000001',beginDate='20050101',endDate=now,field=[],pandas="1")
close=data['closeIndex'].values
turnoverValue=data['turnoverValue'].values
date=data['tradeDate'].values
distance=[]
begindate=[]
enddate=[]
T1=[]
T2=[]
T3=[]
T4=[]
T5=[]
#样本
x=close[-20:]
t=normalized(x)
tv=normalized(turnoverValue[-20:])
nowbegin=date[-20]
nowend=date[-1]
for i in range(len(close)-24):
    begindate.append(date[i])
    enddate.append(date[i+19])
    T1.append((close[i+20]-close[i+19])/close[i+19])
    T2.append((close[i+21]-close[i+20])/close[i+20])
    T3.append((close[i+22]-close[i+21])/close[i+21])
    T4.append((close[i+23]-close[i+22])/close[i+22])
    T5.append((close[i+24]-close[i+23])/close[i+23])
    #比较sample
    r=normalized(close[i:i+20])
    rv=normalized(turnoverValue[i:i+20])
    distance.append(dtw(t,r)+dtw(tv,rv))
#记录
a=pd.DataFrame(begindate,columns=['begindate'])
b=pd.DataFrame(enddate,columns=['enddate'])
w=[distance,T1,T2,T3,T4,T5]
w=np.transpose(w)
c=pd.DataFrame(w,columns=['distance','T1','T2','T3','T4','T5'])
a=a.join(b,how='outer')
record=a.join(c,how='outer')
record=record.sort(columns='distance')
print 'Begindate:  %s' %nowbegin
print 'Enddate:    %s' %nowend
print 'distance前五'
record.head()
```

```python
#我们选前五画图
top5=record.head()
price=np.zeros((6,25))
normprice=np.zeros((6,25))
initialA= np.empty((1,25,))
initialA[:] = np.NAN
norminitialA=np.empty((1,25,))
norminitialA[:]=np.NAN
nclose=normalized(close[-20:])
for i in range(20):
    initialA[0][i]=close[-20+i]
    norminitialA[0][i]=nclose[-20+i]
for i in range(len(top5)):
    day=top5['begindate'].values[i]
    ex=DataAPI.MktIdxdGet(ticker='000001',beginDate=day,endDate=now,field=[],pandas="1")
    normtake=normalized(ex['closeIndex'].values[:25])
    take=ex['closeIndex'].values[:25]
    price[i]=take
    normprice[i]=normtake
price[-1]=initialA
normprice[-1]=norminitialA
price=np.transpose(price)
normprice=np.transpose(normprice)
price=pd.DataFrame(price,index=range(1,26),columns=['Top1','Top2','Top3','Top4','Top5','close'])
normprice=pd.DataFrame(normprice,index=range(1,26),columns=['normTop1','normTop2','normTop3','normTop4','normTop5','normclose'])
price
```
