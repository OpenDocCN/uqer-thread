# 股社区 - 绩优+低估值股票筛选

```python
import numpy as np
import pandas as pd

def getstockIdList(theDay):
    thisYear = theDay[0:4] 
    lastYear = str(int(thisYear)-1)
    lastoneYear = str(int(thisYear)-2)
    lasttwoYear = str(int(thisYear)-3)

    pd.set_option('display.max_rows', 100)
    pd.set_option('colheader_justify', 'right')
    # 2014,2015年报数据
    data = DataAPI.FdmtIndiRtnPitGet(ticker=u"",beginYear=lasttwoYear,endYear=lastoneYear,reportType=u"A",field=u"secID,secShortName,endDate,publishDate,ROE,ROEA")
    data = data.sort_values('publishDate').groupby(['secID','endDate']).last().reset_index()
    data = data.set_index(['secID','secShortName','endDate'])['ROEA'].unstack().reset_index()

    # 2016三季报数据
    q3_data = DataAPI.FdmtIndiRtnPitGet(ticker=u"",beginYear=lastYear,endYear=lastYear,reportType=u"CQ3",field=u"secID,secShortName,endDate,publishDate,ROE,ROEA")
    q3_data = q3_data.sort_values('publishDate').groupby(['secID','endDate']).last().reset_index()
    q3_data = q3_data.set_index(['secID','secShortName','endDate'])['ROEA'].unstack().reset_index()
    data = data.merge(q3_data, how='inner', on=['secID','secShortName'])

    # 市盈率市值数据
    mkt_data = DataAPI.MktStockFactorsOneDayGet(tradeDate=theDay,field='secID,PE,LCAP,LFLO')
    print theDay
    mkt_data['LCAP'] = [np.exp(x)/1e8 for x in mkt_data['LCAP']]
    mkt_data['LFLO'] = [np.exp(x)/1e8 for x in mkt_data['LFLO']]
    data = data.merge(mkt_data, how='inner', on=['secID'])

    # 上市日期数据
    info_data = DataAPI.EquGet(equTypeCD='', secID=data.secID.values, field='secID,listDate', pandas='1')
    data = info_data.merge(data, how='inner', on=['secID'])

    # 行业分类数据
    indu_data = DataAPI.EquIndustryGet(industryVersionCD=u"010303",secID=data.secID.values,
                                       intoDate=theDay,field=u"secID,industryName1",pandas="1")
    data = data.merge(indu_data, how='inner', on=['secID'])

    # 三年平均ROE除以PE
    data.columns = [u'secID', u'listDate', u'secShortName', u'2014_ROE', u'2015_ROE', u'2016Q3_ROE', 
                u'市盈率', u'总市值(亿)', u'流通市值(亿)', u"申万行业分类"]
    # 按ROE筛选
    data = data[data['2014_ROE']>=20][data['2015_ROE']>=20][data['2016Q3_ROE']>=15]
    # 上市时间筛选
    data = data[data['listDate']<='2016-02-14']
    data = data[data[u'总市值(亿)']<=200]
    data[u'三年平均ROE除以PE'] = (data['2014_ROE'] + data['2015_ROE'] + data['2016Q3_ROE'] * 4.0/3) / 3.0 / data[u'市盈率']
    data = np.round(data, 2)
    data = data.sort_values(u'三年平均ROE除以PE', ascending=False).reset_index().drop('index', axis=1)
    datalist = list(data['secID'])
    return datalist
```
