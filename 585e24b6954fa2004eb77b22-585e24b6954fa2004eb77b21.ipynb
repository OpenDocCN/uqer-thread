# 筛选基金

```python
#本代码用于筛选基金
import numpy as np
import pandas as pd

#混合型基金配置比较灵活，更适合进行配置
allFund = DataAPI.FundGet(etfLof=u"",secID=u"",ticker=u"",category=u"H",idxID=u"",idxTicker=u"",listStatusCd=u"",operationMode=u"",field=u"",pandas="1")#E-股票型，H-混合型，B-债券型，SB-短期理财债券，M-货币型，O-其他。
allFund.dropna()

#NAV使用调整后的净值对比，越高越好
#STDEV为标准差，越低标识波动性越小
#sharperatio为夏普比率，也称为性价比，越高越好
df = pd.DataFrame({"ticker":"","nav":"","stdev":"","sharperatio":""},index=["0"])
for i in range(len(allFund)):
    onefund = DataAPI.FundNavGet(dataDate=u"",secID=u"",ticker=allFund.ticker[i],beginDate=u"20140101",endDate=u"20161031",field=u"",pandas="1")
    if len(onefund) == 0 or len(onefund) < 200:
        continue
    nav = onefund.ADJUST_NAV[len(onefund) - 1] - onefund.ADJUST_NAV[0]
    stdev = np.std(onefund['ADJUST_NAV'])
    sharperatio = (nav - 0.1) / stdev
    df.loc[i,['ticker','nav','stdev','sharperatio']] = [onefund.ticker[0],nav,stdev,sharperatio]
    #df.append(res, ignore_index=True)
    
print(df.sort('sharperatio',ascending=0))
print(df[df.stdev<0.3].sort('sharperatio',ascending=0))
```
