### 平稳过程 (Stationary Process)

如果一类过程的主要性质与时间变量的间隔有关，而与所考察的起始点无关，可以称这样的过程为平稳过程。

其严格定义为：

如果随机过程 $\{X(t), t\in T\}$ 对 $\forall t_1, t_2, ..., t_n \in T$ 和 $\forall h \  \text{s.t.} \ t_i + h \in T, i = 1,2,...,n$ 有: $(X(t_1+h), X(t_2+h),..., X(t_n+h))$ 与 $(X(t_1), X(t_2), ..., X(t_n))$ 具有相同的联合分布，记为

$$(X(t_1+h), X(t_2+h),..., X(t_n+h)) =^d= (X(t_1), X(t_2), ..., X(t_n)) $$

则称 $X(t), t\in T$ 为**严平稳**的。

简言之，对严平稳而言，**有限维分布关于时间平移不变**。实际上，严平稳条件过强不易验证，以下引入宽平稳过程。

---
如果随机过程 $\{X(t)\}$ 的所有二阶矩存在，且 $E[X(t)] = \mu$ 与时间无关，协方差函数 $\gamma(t,s)$ 只与时间间隔 $s-t$ 有关，则称此过程为**宽平稳过程**（或**二阶平稳过程、协方差平稳过程**）。以下我们所称的平稳都是指宽平稳。


下面我们生成的序列 $A$ 中的每个点都服从标准正态分布，$\mu = 0$；且每个点互相独立，即 

$Cov(A_m, A_n) = \delta_{mn}$

所以 $A$ 是一个宽平稳序列（具体而言，$A$ 是一个**白噪声序列**)。

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

np.random.seed(107)
T = 200
# 序列 A 中的每个点都服从标准正态分布；其均值函数是0，与时间无关
A = np.array([np.random.normal(0, 1) for t in range(T)])
# meanOfA 是序列A的时间平均值
meanOfA = np.mean(A)

plt.plot(A)
# 红线是 T 时间内的平均值
plt.hlines(meanOfA, 0, T, colors='r', alpha = 0.5, linewidth = 5)
# 绿线是均值函数
plt.hlines(0, 0, T, colors='g', alpha = 0.5, linewidth = 10)
plt.xlabel('Time')
plt.ylabel('Value')
```

从上图我们可以看出**均值回归**的性质：如果满足遍历性条件，当 Value 高于其均值函数 $\mu$，我们可以预期其在将来某时会下降，回归到均值；如果低于均值函数 $\mu$，我们可以预期其会上升，回归到均值。

```python
# 序列 B 的均值函数是 0.1 * t
B = np.array([np.random.normal(0.1 * t, 1) for t in range(T)])
# meanOfB 是序列 B 的时间平均值
meanOfB = np.mean(B)

plt.plot(B)
plt.hlines(meanOfB, 0, T, colors='r')
plt.plot(np.array([0.1 * t for t in range(T)]), c='g')
plt.xlabel('Time')
plt.ylabel('Value')
```

但是对于非平稳过程，长程时间均值将难以用于预测未来。

### 平稳性统计检验 (Testing for Stationarity)

```python
from statsmodels.tsa.stattools import adfuller

def check_for_stationarity(X, cutoff = 0.05):
    pvalue = adfuller(X)[1]
    if pvalue <= cutoff:
        print("p-value:%f"%pvalue, "This series is likely stationary")
    else:
        print("p-value:%f"%pvalue, "This series is likely not stationary")
```

```python
check_for_stationarity(A)
```

```python
check_for_stationarity(B)
```

### 单整阶数 (Order of Integration)

给定一个非平稳过程，如果对其做 $n$ 次差分可以得到平稳过程，就称其为 $I(n)$ 过程。 

我们用 $I(0)$ 表示零阶单整阶数，$I(1), I(2), ..., I(n)$ 以此类推。

反过来，对 $I(0)$ 累积求和可以得到 $I(1)$，对 $I(1)$ 累积求和得到 $I(2)$……

```python
A1 = np.cumsum(A)
A2 = np.cumsum(A1)

plt.subplot(3,1,1)
plt.plot(A)
plt.subplot(3,1,2)
plt.plot(A1)
plt.subplot(3,1,3)
plt.plot(A2)
```

### 协整 (Cointegration)

我们提到过对于非平稳过程，不能期望其均值回归。但是多个非平稳过程的线性组合有可能是平稳的呀！

```python
beginDate = "2013-01-01"
endDate = "2016-01-01"

data1 = DataAPI.MktEqudAdjGet(tradeDate=u"",secID=u"",ticker=u"601169",beginDate=beginDate,endDate=endDate,isOpen="",field=u"closePrice",pandas="1")
data2 = DataAPI.MktEqudAdjGet(tradeDate=u"",secID=u"",ticker=u"601328",beginDate=beginDate,endDate=endDate,isOpen="",field=u"closePrice",pandas="1")
plt.plot(data1, label = 601169)
plt.plot(data2, label = 601328)
plt.legend()
plt.xlabel('Time')
plt.ylabel('Value')
```

```python
Data1 = data1.values.ravel()
Data2 = data2.values.ravel()
beta, alpha = np.polyfit(Data1, Data2, 1)
plt.scatter(Data1, Data2, edgecolor = 'none')
plt.plot(Data1, beta * Data1 + alpha, c = 'r')
plt.xlabel('601169')
plt.ylabel('601328')
```

```python
plt.plot(beta * Data1 + alpha)
plt.plot(Data2)
plt.xlabel('Time')
plt.ylabel('Value')
```

```python
residual = Data2 - (beta * Data1 + alpha)
plt.plot(residual)
plt.hlines(np.mean(residual), data2.index[0], data2.index[-1], colors = 'r', label = 'mean')
plt.hlines(np.mean(residual) + np.std(residual), data2.index[0], data2.index[-1], colors = 'g', label = 'mean + std')
plt.hlines(np.mean(residual) - np.std(residual), data2.index[0], data2.index[-1], colors = 'black', label = 'mean - std')
plt.xlabel('Time')
plt.ylabel('Value')
plt.legend()
```

```python
check_for_stationarity(residual)
```

```python
plt.plot(residual)
plt.plot(pd.Series(residual).rolling(window = 30).mean())
plt.plot(pd.Series(residual).rolling(window = 30).mean() + 1 * pd.Series(residual).rolling(window = 30).std())
plt.plot(pd.Series(residual).rolling(window = 30).mean() - 1 * pd.Series(residual).rolling(window = 30).std())
plt.xlabel('Time')
plt.ylabel('Value')
```
