- ##### 经典的资产配置方式就是马格维茨的均值-方差模型。目标是在给定预期收益率下最小化方差(风险)，或给定风险水平下最大化收益，通过拉格朗日乘子法，可以计算出一个有效前沿，我们可以根据有效前沿来配置资产。但在实践过程中，我们常常发现计算的结果是某几个资产的权重特别大，收益和风险都集中在了这些资产上。

```python
#加载要用到的包
import numpy as np
import pandas as pd
import math
from math import *

import matplotlib as mpl
from matplotlib import pyplot as plt
mpl.style.use('ggplot')
from CAL.PyCAL import *
font.set_size(22)
small_font = font.copy()
small_font.set_size(14)
figsize = (14,14*0.518)
```

- #### 20140101-20151126的基金净值(复权数据)的日涨跌数据作为训练集，比如用于计算协方差阵。

```python
startdate = '20140101'
enddate = '20151126'

ticker = ['000342','096001','000216','040046','160416']
rtns = pd.DataFrame()   #用于存储日收益率序列
for i in range(len(ticker)):
    data = DataAPI.FundNavGet(ticker=ticker[i],beginDate=startdate,endDate=enddate,field=u"secShortName,endDate,ADJUST_NAV",pandas="1")
    tmp = pd.DataFrame()
    tmp['endDate'] = data['endDate'][1:]
    tmp[data.iat[0,0]] = data['ADJUST_NAV'].values[1:]*1.0/data['ADJUST_NAV'].values[:len(data)-1]-1     #得到当日收益率
    tmp.set_index('endDate',inplace=True)
    rtns = pd.concat([rtns,tmp],axis=1)
rtns.fillna(0, inplace = True)  #  将NaN置换为0
```

- #### 20151127-20160805的基金净值(复权数据)的日涨跌数据作为测试集，用于观察在这段时间内的投资组合的表现

```python
startdate = '20151126'
enddate = '20160805'

ticker = ['000342','096001','000216','040046','160416']
rtn_test = pd.DataFrame()   #用于存储日收益率序列
for i in range(len(ticker)):
    data = DataAPI.FundNavGet(ticker=ticker[i],beginDate=startdate,endDate=enddate,field=u"secShortName,endDate,ADJUST_NAV",pandas="1")
    tmp = pd.DataFrame()
    tmp['endDate'] = data['endDate'][1:]
    tmp[data.iat[0,0]] = data['ADJUST_NAV'].values[1:]*1.0/data['ADJUST_NAV'].values[:len(data)-1]-1     #得到当日收益率
    tmp.set_index('endDate',inplace=True)
    rtn_test = pd.concat([rtn_test,tmp],axis=1)
rtn_test.fillna(0, inplace = True)  #  将NaN置换为0
```

```python
rtn_test.head()
```

- #### 先定义计算衡量投资组合表现的各个指标的函数，计算公式可以参考[API文档:风险指标](https://uqer.datayes.com/help/faqApi/#风险指标)：

```python
#从日收益率序列得到累积收益率序列
def get_accumReturns(daily):
    accumReturns=np.zeros(len(daily))
    for i in range(len(daily)):
        tmp = 1.0
        for j in range(i+1):
            tmp *= (1+daily[j])
        accumReturns[i] = tmp-1.0
    return accumReturns
```

```python
#得到年化收益率
def get_annualReturns(accumReturns):
    annualReturns = (1+accumReturns[-1])**(250.0/len(accumReturns))-1
    return annualReturns
```

```python
#得到年波动率
def get_volatility(daily):
    volatility = daily.std()*sqrt(250)
    return volatility
```

```python
#得到夏普比率，rf取3.5%
def get_sharpe(annualReturns,volatility):
    rf = 3.5/100
    sharpe = (annualReturns-rf)/volatility
    return sharpe
```

```python
#得到最大回撤
def get_maxDrawdown(accumReturns):
    drawdown = np.zeros(len(accumReturns))
    for i in range(1,len(drawdown)):
        for j in range(i):
            tmp = np.zeros(i)
            tmp[j] = 1.0-(1+accumReturns[j])/(1+accumReturns[i])
        drawdown[i] = max(tmp)
    maxDrawdown = max(drawdown)
    return maxDrawdown
```

```python
#得到Calmar比率
def get_calmarRatio(annualReturns,maxDrawdown):
    calmarRatio = annualReturns/maxDrawdown
    return calmarRatio
```

- #### 先看一下各个资产自身在20151127-20160805这段时间里表现如何：

```python
#计算各资产自身的风险指标
result1 = pd.DataFrame(np.zeros((6,5)))
accumReturn = pd.DataFrame()
for i in range(5):
    Returns = np.array(rtn_test.iloc[:,[i]]).reshape(len(rtn_test),)#得到资产的收益率序列
    accumReturns = get_accumReturns(Returns)    #调用函数计算各项指标
    accumReturn[str(i)] = accumReturns
    annualReturns = get_annualReturns(accumReturns)
    volatility = get_volatility(Returns)
    sharpe = get_sharpe(annualReturns,volatility)
    maxDrawdown = get_maxDrawdown(accumReturns)
    calmarRatio = get_calmarRatio(annualReturns,maxDrawdown)
    result1.iat[0,i] = accumReturns[-1]
    result1.iat[1,i] = annualReturns   #把结果存进dataframe里
    result1.iat[2,i] = volatility
    result1.iat[3,i] = sharpe
    result1.iat[4,i] = maxDrawdown
    result1.iat[5,i] = calmarRatio
accumReturn.columns = rtns.columns
accumReturn.index = np.array(rtn_test.index)
result1.columns = rtns.columns
result1.index = ['累计收益率','年化收益率','波动率','夏普比率','最大回撤','Calmar比率']
result1
```

- #### 画个图更加直观：

```python
#来自涛哥的中文画图函数
def fig_style(ax, legend, legend_loc='upper right', xlabel=''):
    """自定义图形属性及格式：
    ax: 图形对象
    legend: 标注
    legend_loc: 标注位置
    xlable: x 坐标名称
    """
    ax.legend(legend, prop=small_font, loc=legend_loc)
    ax.title.set_font_properties(font)
    if xlabel:
        ax.set_xlabel(xlabel, fontproperties=small_font)
    ax.set_axis_bgcolor('white')
    ax.grid(color='gray', alpha=0.2, axis='y')
    return ax
```

# ** 三.各种资产配置组合 **
## **1.蛋卷安睡全天候(海外)的投资组合**
- #### 我在网上找到了蛋卷安睡全天候(海外)开售时所公布的各个基金的配比：

```python
weight = np.array([0.55,0.2,0.1,0.1,0.05])   #权重序列
weights1 = pd.DataFrame()
weights1['权重'] = weight
weights1.index = rtns.columns
weights1
```

- ### 可以看到，债券的配比非常高，极大程度上降低了整个投资组合的风险，石油的比重非常小，更加规避了波动大的资产；
- ### 这段时间大涨的黄金并没有占太大的比重，导致整个投资组合的收益主要来源于债券，这将使投资组合收益比债券高的不多。
- ### 我们来看看该投资组合的表现：

```python
#定义一个计算投资组合风险指标的函数
def get_result(weight):
    Returns = np.mat(rtn_test.values)*np.mat(weight).T               #等权组合的收益率
    Returns = np.array(Returns).reshape(len(Returns),)       #把等权组合的收益率序列处理成数组

    #计算风险指标
    result2 = pd.DataFrame()
    accumReturns = get_accumReturns(Returns)    #调用函数计算各项指标
    annualReturns = get_annualReturns(accumReturns)
    volatility = get_volatility(Returns)
    sharpe = get_sharpe(annualReturns,volatility)
    maxDrawdown = get_maxDrawdown(accumReturns)
    calmarRatio = get_calmarRatio(annualReturns,maxDrawdown)
    result2['等权组合'] = [accumReturns[-1],annualReturns,volatility,sharpe,maxDrawdown,calmarRatio]  #把结果存进dataframe里
    result2.index = ['累计收益','年化收益率','波动率','夏普比率','最大回撤','Calmar比率']
    return result2,accumReturns
```

```python
result1,accumReturns1 = get_result(weight)
result1.columns = ['蛋卷安睡全天候(海外)']
result1
```

- ### 正如我们所推测的，该投资组合的累积收益也就比债券高一些，波动和回撤被控制在了一个比较小的范围，之后我们再分析这个投资组合在分散风险方面的优劣。
- ### 先来看看其它组合方式的情况。

## **2.等权重投资组合**

- ### 个人认为，不管在何种情况下，等权重的方式总不是最差的投资组合，我们来看看其表现：

```python
weight = np.array([1.0/5 for i in range(5)])   #权重序列
weights2 = pd.DataFrame()
weights2['权重'] = weight
weights2.index = rtns.columns
print weights2.to_html()
result2,accumReturns2 = get_result(weight)   #获取表现
result2.columns = ['等权组合']
print result2.to_html()
```

- ### 收益率提高了一些，但是波动也变大了，这个结果可以和之后的组合进行比较。

## **3.波动率倒数加权组合**

- ### 按照分散风险的思想，我们期望的是各个资产在投资组合中的风险（波动率）都一样，这样才能使得风险不会暴露在某一或某几个资产上
- ### 这里的波动率的计算使用的是20140101-20151126的数据

```python
#计算波动率倒数加权的各项资产权重
w = 1.0/rtns.std()   #各项资产波动率倒数
weight = np.array([w[i]/w.sum() for i in range(5)])  #波动率倒数加权，保证权重为1
weights3 = pd.DataFrame()
weights3['权重'] = weight
weights3.index = rtns.columns
print weights3.to_html()
result3,accumReturns3 = get_result(weight)   #获取表现
result3.columns = ['波动率倒数加权']
print result3.to_html()
```

- ### 和等权重组合相比，该组合在收益率提升了的同时，波动率下降了，但还是不能和蛋卷组合的小波动率相比

# **4.VaR组合**

- ### 关于VaR的基础知识，可以参考[在险价值(Value at Risk)风险度量下的资产组合前沿](https://uqer.datayes.com/community/share/57aa83f8228e5b9b95a88894)
- ### 计算权重使用的是20140101-20151126的数据,$\alpha$用的是0.01

```python
import scipy.optimize as opt 

riskLevel = 0.01

def histVaR(histSeries, riskLevel = 0.01):
    return -histSeries.describe(percentiles = [riskLevel])[str(int(riskLevel * 100))+'%']
def pfVaR(weights):
    return histVaR(pd.DataFrame(np.mat(rtns)*np.mat(weights).T, columns = ['portVaR']).portVaR, riskLevel= riskLevel)

cons = ({'type': 'eq', 'fun': lambda x: sum(x) - 1})
res = opt.minimize(pfVaR, [1/5 for i in range(5)], constraints= cons, bounds= [(0,1)]*5, method = 'SLSQP')
```

```python
weight = res['x']
weights4 = pd.DataFrame()
weights4['权重'] = weight
weights4.index=rtns.columns
print weights4.to_html()
result4,accumReturns4 = get_result(weight)   #获取表现
result4.columns = ['最小VaR']
print result4.to_html()
```

- ### 可以看到收益大幅提升的同时,波动也得到了一定程度的控制。
