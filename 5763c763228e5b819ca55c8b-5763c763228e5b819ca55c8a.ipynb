## 前言

对于多因子选股很多人都有所了解,在合成信号的时候,我们需要对各因子分配权重。一般有两种方法------- 打分法和回归法。<br>
打分法就是自行确定每个因子的系数，例如[量化分析师的Python日记【第14天：如何在优矿上做Alpha对冲模型】](https://uqer.datayes.com/community/share/55e662f9f9f06c1ea481f9cf)中，对因子NetProfitGrowRate、ROE、RSI进行了打分法<br>
代码第60行:weight = np.array([0.4, 0.3, 0.3])   #　信号合成，各因子权重<br>

而回归法就是利用历史数据去拟合而得到拟合系数作为每个因子的权重，本帖研究的内容就算比较最小二乘回归与lasso回归在多因子权重分配上的效果。

关于lasso的介绍可参阅[http://bbs.pinggu.org/thread-3848519-1-1.html](http://bbs.pinggu.org/thread-3848519-1-1.html)

## 一、数据准备
主要有以下几点：<br>
1.观察样本主体为HS300股票池中的股票，每个月的因子值，回归目标为该月的月收益率<br>
2.为保证不使用未来数据，我们取20110801-20130531的月数据用来训练模型，20130601-20160601进行回测。<br>
3.这里选了8个因子构建多因子模型，也可以自行选择其他因子。(PE、PB、ROE、LFLO、20日资金流量、应付账款周转率、净利润增长率、FiftyTwoWeekHigh(当前价格处于过去1 年股价的位置))<br>
4.对数据进行清洗、去极值、中性化、标准化、归一化<br>
代码如下:

1.取数据，代码运行时间略长，一次运行后存为csv再进行读取，避免重复运行。在取数据的过程中完成对数据的处理操作

```python
## 代码运行时间略长，一次运行后存为csv
import pandas as pd

def CalcNextDate(date,date_list):            ## 得到下一个月日期
    for i in range(len(date_list)-1):
        if date==date_list[i]:
            return date_list[i+1]

universe = set_universe('HS300')
data=DataAPI.TradeCalGet(exchangeCD=u"XSHE",beginDate=u"20110801",endDate=u"20130531",field=['calendarDate','isMonthEnd'],pandas="1")   ## 得到日期列表
data = data[data['isMonthEnd'] == 1]
date_list_ext = data['calendarDate'].values.tolist()
date_list = date_list_ext[:-1]  ## 最后一个月无法计算下个月的收益 

df = pd.DataFrame()

## 获取因子数据
for date in date_list:
    temp = DataAPI.MktStockFactorsOneDayGet(tradeDate=date,secID=universe,field=u"secID,tradeDate,PE,LFLO,NetProfitGrowRate,AccountsPayablesTRate,ROE,MoneyFlow20,PB,FiftyTwoWeekHigh",pandas="1")
    temp = temp.dropna()
    temp = temp.set_index(temp['secID'])
    date_t = date[:4]+date[5:7]+date[8:]
    temp['PE'] = pd.Series(standardize(neutralize(winsorize(temp['PE'].to_dict(),n_draw=10),date_t)))
    temp['LFLO'] = pd.Series(standardize(neutralize(winsorize(temp['LFLO'].to_dict(),n_draw=10),date_t)))
    temp['NetProfitGrowRate'] = pd.Series(standardize(neutralize(winsorize(temp['NetProfitGrowRate'].to_dict(),n_draw=10),date_t)))
    temp['MoneyFlow20'] = pd.Series(standardize(neutralize(winsorize(temp['MoneyFlow20'].to_dict(),n_draw=10),date_t)))
    temp['PB'] = pd.Series(standardize(neutralize(winsorize(temp['PB'].to_dict(),n_draw=10),date_t)))
    temp['AccountsPayablesTRate'] = pd.Series(standardize(neutralize(winsorize(temp['AccountsPayablesTRate'].to_dict(),n_draw=10),date_t)))
    temp['ROE'] = pd.Series(standardize(neutralize(winsorize(temp['ROE'].to_dict(),n_draw=10),date_t)))
    temp['FiftyTwoWeekHigh'] = pd.Series(standardize(neutralize(winsorize(temp['FiftyTwoWeekHigh'].to_dict(),n_draw=10),date_t)))                                   
    df = pd.concat([df,temp])

    
df['Gain']=pd.Series('x',index=df.index)               ##  月收益率
length = len(df)
newindex=[]
for i in range(length):
    newindex.append(i)
df.index = newindex 

## 获取当月及下月股价，计算月收益率
for i in range(length):
    temp = df[i:i+1]
    temp_secID = df[i:i+1]['secID'].values[0]
    temp_Date = df[i:i+1]['tradeDate'].values[0]
    next_Date = CalcNextDate(temp_Date,date_list_ext)
    try:
        price1 =DataAPI.MktEqudGet(tradeDate=temp_Date,secID=temp_secID,isOpen="",field=['closePrice'],pandas="1")[0:1][['closePrice']].values[0][0]
        price2 = DataAPI.MktEqudGet(tradeDate=next_Date,secID=temp_secID,isOpen="",field=['closePrice'],pandas="1")[0:1][['closePrice']].values[0][0]
        MonthGain = (price2-price1)/price1
    except:
        pass
    df['Gain'][i] = MonthGain  
    
df.to_csv('data_3.csv')
```

2.数据的后续处理，及训练集与测试集的划分。 共4000组数据用于训练，729组数据其实是用来参数寻优的，真正的测试用回测框架进行，后续介绍。<br>
另外，这里也可以用交叉验证的方法训练模型。

```python
import pandas as pd
import numpy as np

# 归一化
def normalization(series):
    return (series - min(series))/(max(series) - min(series)) 

import pandas as pd
df = pd.read_csv("data_3.csv")
del df['Unnamed: 0']
df = df.dropna()

print len(df)
df['Gain'] = 100*normalization(df['Gain'])  # 这里乘以了100，主要是为了方便观察回归系数，否则系数太小。

df['PE'] = normalization(df['PE'])
df['LFLO'] = normalization(df['LFLO'])
df['NetProfitGrowRate'] = normalization(df['NetProfitGrowRate'])
df['MoneyFlow20'] = normalization(df['MoneyFlow20'])
df['PB'] = normalization(df['PB'])
df['AccountsPayablesTRate'] = normalization(df['AccountsPayablesTRate'])
df['ROE'] = normalization(df['ROE'])
df['FiftyTwoWeekHigh'] = normalization(df['FiftyTwoWeekHigh'])

# 矩阵化
rawmat = np.mat(df)
mat = rawmat[:,2:10]
y = rawmat[:,10]
print mat.shape,y.shape

# 数据集划分
trainMat = mat[0:4000] 
testMat = mat[4000:]
trainY = y[0:4000]
testY = y[4000:]
```

## 二、OLS与lasso效果对比

1.先看OLS构建的多因子策略，利用sklearn包进行计算。

```python
from sklearn import linear_model
clf = linear_model.LinearRegression()
clf.fit(trainMat,trainY)
weights_OLS = clf.coef_[0]
# 查看回归系数
weights_OLS
```

根据回归系数构建多因子策略：<br>
1.回测时间完全在样本外: 20130601-20160601,调仓频率为1个月<br>
2.根据调仓日前一天的因子与回归系数的乘积作为分数，取排名前20的股票，按分数大小计算买入权重。

2.可以看出该多因子模型是有效的，但是表现一般。我们再来看看lasso的效果

第一个遇到的问题是lasso参数alpha的选取。这里采用参数寻优的方法，优化标准为预测集与测试集误差的平方和最小。(这里步长似乎太小了，耗时很久，可以调节)<br>
当然也可以自行调节参数，其实这里自行调节参数为0.009以上，回测的效果已然差距不大。

**下个cell的代码可以不运行**

```python
# 参数寻优
best_err = 1000000000
best_lam = 0
for i in range(10000):
    lam = 0.0001*(i+1)
    clf2 = linear_model.Lasso(alpha=lam)
    clf2.fit(trainMat,trainY)
    y_pre = clf2.predict(testMat)
    err = ((np.array(y_pre)-np.array(testY))**2).sum()
    if err < best_err:
        best_lam = lam
        best_err = err
print best_lam   
```

构建模型：

```python
clf2 = linear_model.Lasso(alpha=0.0194)
clf2.fit(trainMat,trainY)
weights_lasso = clf2.coef_
weights_lasso
```

我们可以看出，其他因子系数都为0，而只保留了 FiftyTwoWeekHigh 这一个因子！<br>那就先看看效果吧

可以发现，收益率、夏普比率、信息比率均比OLS的策略效果好！<br>
参数寻优的结果用到测试集上也未必效果最好，我们再调节参数大小试试

```python
clf2 = linear_model.Lasso(alpha=0.0096)
clf2.fit(trainMat,trainY)
weights_lasso = clf2.coef_
weights_lasso
```

```python
clf2 = linear_model.Lasso(alpha=0.009)
clf2.fit(trainMat,trainY)
weights_lasso = clf2.coef_
weights_lasso
```

```python
clf2 = linear_model.Lasso(alpha=0.0075)
clf2.fit(trainMat,trainY)
weights_lasso = clf2.coef_
weights_lasso
```

```python
clf2 = linear_model.Lasso(alpha=0.0072)
clf2.fit(trainMat,trainY)
weights_lasso = clf2.coef_
weights_lasso
```

```python
clf2 = linear_model.Lasso(alpha=0.007)
clf2.fit(trainMat,trainY)
weights_lasso = clf2.coef_
weights_lasso
```

如果将这些回归系数分别放到策略里运行会发现，当参数alpha的值在0.09-0.02（也不能太大）以上，回测的效果差不多，而小于0.09表现会变差（可以自行尝试）<br>

我们发现，改变参数，可以自由的选择那些“更好”的因子，而舍弃不那么重要的因子，这里影响最大的因子为FiftyTwoWeekHigh、其次为PB、再次为LFLO,其他因子的可以略去。这使模型有更好的解释能力，也降低了无用因子的干扰。<br>

可以设想，理论上，我们可以通过参数寻优，从优矿提供的所有因子中选出一定数量表现最好的因子！
