# 高转送logistic预测模型

## 摘要
---
本文以全A股股票为股票池，建立了一个基于logistic回归的高转送预测模型。虽说现在在刘主席的监管下，高转送等炒作可能会逐渐失效。但是没有关系，通过找一些feature建立模型来预测事件发生的可能性，这种思路永远不会过时。倘若高转送不行了，我们可以很容易的切换到预测有党性的高分红事件嘛。

## 获取高转送数据
---
这里我们调用DataAPI，EquDivGet获取历史上各只股票的分红送转情况，参考了兴业证券的研报，将高转送定义为转股比例加上送股比例大于0.5。

```python
import pandas as pd
import numpy as np
from dateutil.parser import parse
import datetime
import matplotlib.pylab as plt
from CAL.PyCAL import *
import statsmodels.api as sm 
from sklearn.linear_model import LogisticRegression
cal = Calendar('China.SSE')
```

```python
def get_universe(trade_date):
    """
    获取动态股票列表
    """
    eq_msg = DataAPI.EquGet(equTypeCD=u"A",secID=u"",ticker=u"",listStatusCD=u"L,S,DE,UN",
               field=u"secID,listDate,delistDate",pandas="1")
    eq_msg = eq_msg.fillna(datetime.datetime.today().strftime('%Y-%m-%d'))
    eq_msg = eq_msg[(eq_msg['listDate']<trade_date) & (eq_msg['delistDate']>trade_date)]
    return eq_msg['secID'].tolist()
```

```python
universe = DataAPI.EquGet(equTypeCD=u"A",secID=u"",ticker=u"",listStatusCD=u"L,S,DE,UN",field=u"secID",pandas="1")['secID'].tolist()#包括退市的股票
```

```python
div_summary = DataAPI.EquDivGet(exDivDate="",eventProcessCD=u"",secID=universe,ticker=u"",
                  beginDate=u"",endDate=u"",field=u"secID,publishDate,endDate,perShareDivRatio,perShareTransRatio",pandas="1")
div_summary = div_summary.dropna(subset=['publishDate'])
div_summary = div_summary.fillna(0)
div_summary['TotalRatio'] = div_summary['perShareDivRatio'] + div_summary['perShareTransRatio'] #转股比例加上送股比例大于0.5
high_amount = div_summary[div_summary['TotalRatio'] > 0.5].reset_index(drop=True)
high_amount['year'] = high_amount['publishDate'].map(lambda x:x[0:4])
```

```python
high_amount.groupby(['year'])['secID'].count().plot(kind='bar',figsize = (15,8))
```

如上便是各年份的发生高送转的数量，我国07年会计准则发生过大的改变，我们选取07年至16年作为我们的样本区间。

## 建立模型
---
目标：我们通过11月份已知的信息来预测当年年终该股票是否会发生高转送。
### 考虑变量(信息)
* 每股资本公积金与留存收益之和
* EPS(TTM)
* 总股本
* 20个交易日平均股价
* 是否新股(上市250个交易日以内)
* 上市时间(年)
* 年中是否有分红
* 去年是否有高转送

```python
def get_point_date(start,end):
    point_date = []
    for year in range(start,end):
        point_date.append(str(year)+'-11-01')
    return point_date
def get_point_universe(point_date):
    available_universe_dict = {}
    for date in point_date:
        available_universe_dict[date] = get_universe(date)
    return available_universe_dict
point_date = get_point_date(2007,2015) ## 得到2007年11月1至2014年11月1日
universe_dict = get_point_universe(point_date)
```

```python
def get_factor(trade_date,universe):
    trade_date = cal.adjustDate(trade_date).strftime('%Y%m%d')
    begin_date = cal.advanceDate(trade_date,'-20B').strftime('%Y%m%d')
    
    ##得到每股资本公积金与每股留存收益之和与EPS
    fundament = DataAPI.MktStockFactorsOneDayProGet(tradeDate=trade_date,secID=universe,ticker=u"",
                                    field=u"secID,tradeDate,CapitalSurplusFundPS,RetainedEarningsPS,EPS",pandas="1")
    fundament['Total_PS'] = fundament['CapitalSurplusFundPS'] + fundament['RetainedEarningsPS'] # 每股资本公积金与每股留存收益之和
    ##得到总股本
    quotation = DataAPI.MktEqudGet(tradeDate=trade_date,secID=universe,ticker=u"",
                                   beginDate=u"",endDate=u"",isOpen="",field=u"secID,closePrice,marketValue",pandas="1")
    quotation['Total_Shares'] = quotation['marketValue'] / quotation['closePrice'] / 1e+08 ##总股本(亿股)
    
    ##计算平均股价
    quotation_window = DataAPI.MktEqudGet(tradeDate=u"",secID=universe,ticker=u"",
                                   beginDate=begin_date,endDate=trade_date,isOpen="",field=u"secID,closePrice",pandas="1")
    # print quotation_window,begin_date,trade_date
    quotation_window = quotation_window.groupby(['secID']).mean().reset_index().rename(columns = {'closePrice':'meanPrice'})
    
    ##计算上市时间
    eq_list = DataAPI.EquGet(equTypeCD=u"A",secID=universe,ticker=u"",listStatusCD=u"",field=u"secID,listDate",pandas="1")
    eq_list['listYear'] = eq_list['listDate'].map(lambda x: cal.bizDatesNumber(x,trade_date)/250.0) 
    
    ##设置是否为次新
    eq_list['ifNew'] = eq_list['listYear'].map(lambda x:1 if x<=1 else 0)
    
    summary_df = pd.merge(fundament[['secID','tradeDate','Total_PS','EPS']],quotation[['secID','Total_Shares']],on=['secID'],how='inner')
    summary_df = pd.merge(summary_df,eq_list[['secID','listYear','ifNew']],on=['secID'],how='inner')
    ##年中是否有过高转送
    mid_transfer_list = high_amount[(high_amount['secID'].isin(universe)) & (high_amount['endDate']== (trade_date[0:4]+'-06-30'))]['secID'].tolist()
    summary_df['MidHasTransfer'] = summary_df['secID'].map(lambda x: 1 if x in mid_transfer_list else 0)
    ##去年是否有过高转送
    last_transfer_list = high_amount[(high_amount['secID'].isin(universe)) & (high_amount['endDate']== str(int(trade_date[0:4])-1)+'-12-31')]['secID'].tolist()
    summary_df['LastHasTransfer'] = summary_df['secID'].map(lambda x: 1 if x in last_transfer_list else 0)
    
    return summary_df
```

```python
def get_total_summary(universe_dict):
    """
    获取所有样本点汇总
    """
    total_summary_df = pd.DataFrame()
    for k,v in universe_dict.items():
        summary_df = get_factor(k,v)
        result = high_amount[(high_amount['publishDate']>k) & (high_amount['publishDate']<str(int(k[0:4])+1)+'-05-01')]
        summary_df['isTransfer'] = summary_df['secID'].map(lambda x:1 if x in result['secID'].tolist() else 0)
        total_summary_df = pd.concat([total_summary_df,summary_df],axis=0)
    return total_summary_df
```

汇总所有变量

```python
total_summary_df = get_total_summary(universe_dict)
total_summary_df = total_summary_df.dropna()
total_summary_df = total_summary_df.reset_index(drop=True)
```

得到解释变量与被解释变量构成的矩阵，其中`Total_PS`是每股资本公积金与每股留存收益之和，`EPS`是TTM的每股收益、`Total_Shares`是总股本单位为亿、`listYear`是上市时间单位年、`ifNew`是虚拟变量表示是否是新股、`MidHasTransfer`是虚拟变量表示年中是否有过高转送、`LastHasTransfer`是虚拟变量表示去年是否有过高转送，一共7个解释变量；`isTransfer`是被解释变量表示年终是否真的高转送。

```python
total_summary_df.head()
```

```python
total_summary_df['intercept'] = 1.0##加入常数项
total_summary_df.head()
```

```python
trans_col = list(total_summary_df.columns[2:-2]) + [total_summary_df.columns[-1]] #获取解释变量列
print trans_col,
```

```python
logit = sm.Logit(total_summary_df['isTransfer'],total_summary_df[trans_col])
result = logit.fit()
# classifier = LogisticRegression()  # 使用类，参数全是默认的  
# classifier.fit(total_summary_df[trans_col], total_summary_df['isTransfer'])  # 训练数据来学习，不需要返回值  
```

```python
print result.summary()
```

最终的回归结果如上所示，可以看出Total_PS（每股资本公积金与每股留存收益之和）与EPS越大，年终进行高转送的概率也越大，并且显著性非常高；Total_Shares（总股本）和listYear（上市时间）越小，年终进行高转送的概率也越大，并且显著性非常高；而ifNew（是否是新股）系数是正的，但是不显著性，MidHasTransfer（年中是否高转送）也不显著；LastHasTransfer（去年是否高转送）比较显著。

## 样本内与样本外预测
---
依据我们得到的模型进行样本内外预测

```python
in_Sample_Predict = result.predict(total_summary_df[trans_col])
in_Sample_Result = total_summary_df[:]
in_Sample_Result['predict_probability'] = pd.Series([round(prob,3) for prob in in_Sample_Predict],index=in_Sample_Result.index)
in_Sample_Result.groupby(['tradeDate']).apply(lambda x:x.sort_values(['predict_probability'],ascending = False))[['secID','isTransfer','predict_probability']]
```

如上便是所有样本内的情况,大家可以对比其预测高转送概率（predict_probability）与是否发生了高转送（isTransfer）。以2015年年终的高转送实际情况，作为样本外点预测。

```python
out_Sample_Universe = get_universe('2015-11-01')
out_Sample_df = get_factor('2015-11-01',out_Sample_Universe)
out_Sample_df = out_Sample_df.dropna()
out_Sample_df['intercept'] = 1
out_Sample_df.head()
```

2015年样本外解释变量如上所示。

```python
out_Sample_Predict = result.predict(out_Sample_df[trans_col])
out_Sample_Result = out_Sample_df[:]
out_Sample_Result['predict_probability'] = pd.Series([round(prob,3) for prob in out_Sample_Predict],index=out_Sample_Result.index)
out_Sample_Result = out_Sample_Result.sort_values(['predict_probability'],ascending=False)
out_Sample_Realized = high_amount[high_amount['endDate']=='2015-12-31']
out_Sample_Result['isTransfer'] = out_Sample_Result['secID'].map(lambda x:1 if x in out_Sample_Realized['secID'].tolist() else 0)
```
