# 300EFT套利机会估算

基于收盘价的折价率数据判断套利机会，基于隔日开盘价买入，这种沪深300指数ETF搬砖策略，交易机会太少了。
沪深市场上一共有4只300ETF，目前510300和159912日交易量在数十亿；510310交易量在亿元左右，159919的日交易量只有几百万，流动性太差了。所以，研究前两只ETF基金。
我用2016年初到2019年9月1日的数据做了分析510300和159912之间的折价率之差，判断两者之间的套利机会。
研究表明，交易的机会比较很少。
首先，510300长期比159912折价更多 。894个交易日中的774个交易日，510300折价率-159912折价率<0。搬砖需要低买高卖，在这种情况下，如果不考虑交易成本，滑点，隔日开盘涨跌幅，那么差不多会有120个交易机会。
其次，假设单边交易成本为万2，510300折价率-159912折价率>0.0004的共有76次，510300折价率-159912折价率>-0.0004共有723次。所以，则自不考虑滑点，隔日开盘涨跌幅影响的情况下，超过交易成本的机会，最多有76次。510300折价率-159912折价率>0.001的只有可怜的41次。

```python
start_date='20160101'
last_date='20190901'

sh300_510300df=DataAPI.MktFunddGet(ticker=u"510300",beginDate=start_date,endDate=last_date,field=u"tradeDate,discountRatio",pandas="1")

sh300_510310df=DataAPI.MktFunddGet(ticker=u"510310",beginDate=start_date,endDate=last_date,field=u"discountRatio",pandas="1")

sz300_159919df=DataAPI.MktFunddGet(ticker=u"159919",beginDate=start_date,endDate=last_date,field=u"discountRatio",pandas="1")

sz300_159912df=DataAPI.MktFunddGet(ticker=u"159912",beginDate=start_date,endDate=last_date,field=u"discountRatio",pandas="1")

```
