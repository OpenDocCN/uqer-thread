# 基于多因子选股的二八轮动策略改进

        本文是本人的第一篇分享，希望大家多点赞，多收藏！
        二八轮动策略是一个经典的量化策略，本篇将二八轮动与多因子策略结合，在2011-4-09到2016-07-01段回测数据上看，新策略提高了年化收益率（从44.1%提高到59.5%），夏普比例(从1.51到2.21)和信息比率（1.38到1.83），降低最大回撤（25.2%到21.40%）与收益波动（26.6%到25.1%），同时在市场大跌的2011-04-09到2012-12-31表现也优于原策略：收益率（从-4.1%提高到-1.4%，基准为-15.4%），Alpha（从1.5%提高到3.8%），降低最大回撤（22.5%到21.4%）。显示出在熊市中表现也优于原策略。
        首先构建的是传统意义上的二八轮动策略：
        1.选取沪深300与创业板指标作为换股参考
        2 比较HS300的20日涨幅与CYB的20日涨幅
            2.1 如果HS300的20日涨幅与CYB的20日涨幅都小于0则清空所有股票；
            2.2 如果HS300的20日涨幅大于CYB的20日涨幅则清空所有股票，并买入HS300中流通市值最大的15支股票；
            2.3 如果HS300的20日涨幅小于CYB的20日涨幅则清空所有股票，并买入创业板中流通市值最小的15支股票；

		由于在选股阶段只使用流通市值这个选股因子，相当于一个单因子模型，所以我们可以加入更多的选股因子增加策略的有效性，在这里我们先选取一个分析师推荐评级'REC'，REC与LFLO作为筛选因子重新构建策略：
		1.选取沪深300与创业板指标作为换股参考
		2 比较HS300的20日涨幅与CYB的20日涨幅
			2.1 如果HS300的20日涨幅与CYB的20日涨幅都小于0则清空所有股票；
			2.2 如果HS300的20日涨幅大于CYB的20日涨幅
				2.2.1 则清空所有股票;
				2.2.2 筛选沪深300中流通市值最大的150支股票
				2.2.3 在2.2.2中得到150支股票中取得评级最高的15支股票买入
			2.3 如果HS300的20日涨幅小于CYB的20日涨幅
				2.3.1 则清空所有股票;
				2.3.2 筛选创业板中流通市值最小的66.67%的股票
				2.3.3 在2.3.2中得到的股票中取得评级最高的15支股票买入

		为了降低策略的最大回撤与进一步提高策略的收益率与夏普比例，本文构建一个三因子选股模型,选取流通市值'LFLO',分析师推荐评级'REC'和经营活动产生的现金流量净额与企业价值之比（Cash provided by operations to enterprise value），该因子简写成CFO2EV。以CFO2EV,REC与LFLO作为筛选因子重新构建策略：
		1.选取沪深300与创业板指标作为换股参考
		2 比较HS300的20日涨幅与CYB的20日涨幅
			2.1 如果HS300的20日涨幅与CYB的20日涨幅都小于0则清空所有股票；
			2.2 如果HS300的20日涨幅大于CYB的20日涨幅；
				2.2.1 则清空所有股票；
				2.2.2 筛选沪深300中流通市值最大的200支股票；
				2.2.3 在2.2.2得到的200支股票中筛选CFO2EV最高的100支股票；
				2.2.4 在2.2.3得到的股票中选取分析师评级最高的15支股票等权买入；
			2.3 如果HS300的20日涨幅小于CYB的20日涨幅
				2.3.1 则清空所有股票;
				2.3.2 筛选创业板中流通市值最小的66.67%的股票；
				2.3.3 在2.3.2得到的股票中筛选CFO2EV最高的30.00%的股票（相对于所有创业板股票的30.00%）；
				2.3.4 在2.3.3得到的股票中取得评级最高的15支股票等权买入；

    	可以看到采用三因子选股的二八轮动策略年化收益率达到59.5%远远高于经典的44.4%，这说明多因子选股为策略带来了15.1%的Alpha，同时极大的提高了夏普比例与信息率，最大回撤也减小到了21.4%,这一点远优于双因子选股的二八轮动策略！这说明策略改进的有效。
    	同时在市场大跌的2011-04-09到2012-12-31，多因子选股的二八轮动策略表现也优于经典二八轮动策略：提高了收益率（从-4.1%提高到-1.4%，基准为-15.4%），Alpha（从1.5%提高到3.8%），降低最大回撤（22.5%到21.4%）。读者可以复制代码自行测试。
    	最后读者也可以选取不同的因子，调整筛选比例得到比本文拥有更优的收益率与稳定性的策略！
