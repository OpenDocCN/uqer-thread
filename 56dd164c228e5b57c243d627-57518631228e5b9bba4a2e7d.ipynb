以上代码是dataframe格式数据取 index=‘four’中pop的数据，但是如果要‘state’中名称为'Wsg'的的行‘pop’列数据怎么办，代码如下

```python
import pandas as pd
import numpy as np
from pandas import Series, DataFrame
data={'state':['Ohio','Cesg','Hfsg','Wsg','Jsg'],'year':[2000,2001,2002,2003,2004],'pop':[1.5,1.7,3.6,2.4,2.9]}
frame=DataFrame(data,columns=['state','year','pop'],index=['one','two','three','four','five'])
print(frame)
#eq_EPS_EGRO[eq_EPS_EGRO['secID'] == stk]['Value'].values[0]
stk='Wsg'
frame[frame['state']==stk]['pop'].values[0]
```
