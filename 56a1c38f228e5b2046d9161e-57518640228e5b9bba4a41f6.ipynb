`融资融券`

```python
def rong_zi_rong_quan():
    """融资融券信息
    """
    trade_days = lib.trade_pre_n_days()
    data = DataAPI.FstTotalGet(exchangeCD=u"XSHE,XSHG", beginDate=trade_days[0])
    data.groupby(by='tradeDate').sum()
    data = data.groupby(by='tradeDate').sum()
    data['tradeVal'] = data['tradeVal']/(10**12)
    ax = data.plot(y=['tradeVal'], kind='bar', figsize=(20, 6))
    axhline(y=1.0, color='r')
    
    tradeValChange = [0]
    tmp = data.tradeVal[0]
    for i in data.tradeVal[1:]:
        tradeValChange.append(i - tmp)
        tmp = i
    data['tradeValChange'] = tradeValChange
    data.plot(y=['tradeValChange'], kind='bar', figsize=(20, 6))
    
    return None

rong_zi_rong_quan()
```

`沪港通`

```python
def hu_gang_tong():
    """沪港通数据，沪股通缺失2015-12-24的数据
    
    DataAPI.JY.HKSHKQuotaJYGet(endDate='2015-12-23')
    DataAPI.JY.HKSHKQuotaJYGet(endDate='2015-12-24')
    
    """
    trade_days = lib.trade_pre_n_days()
    res, hushen, hk = pd.DataFrame(), pd.DataFrame(), pd.DataFrame()
    for day in trade_days:
        tmp_1 = DataAPI.JY.HKSHKQuotaJYGet(endDate=day, tradingTypeCd="1")
        tmp_2 = DataAPI.JY.HKSHKQuotaJYGet(endDate=day, tradingTypeCd="2")
        hushen = tmp_1 if hushen.empty else hushen.append(tmp_1) 
        hk = tmp_2 if hk.empty else hk.append(tmp_2) 
    
    hushen['netIn'] = (hushen['dailyQuota'] - hushen['dailiBalance']) / (10 ** 9)
    hk['netIn'] = (hk['dailyQuota'] - hk['dailiBalance']) / (10 ** 9)
    res['tradeDate'] = hushen['endDate']
    res['in_hushen'] = hushen['netIn']
    res['in_hk'] = hk[hk.endDate.isin(hushen.endDate)].netIn

    res.plot(x='tradeDate', y=['in_hushen', 'in_hk'], kind='bar', figsize=(20, 6))
    
    return hushen, hk

hushen, hk = hu_gang_tong()
```

`大盘资金流入流出情况`

```python
def money_flow():
    """资金流入流出情况
    """
    trade_days = lib.trade_pre_n_days()
    data = pd.DataFrame()
    for day in trade_days:
        tmp = DataAPI.MktIndustryFlowOrderGet(tradeDate=day)
        data = tmp if data.empty else data.append(tmp)

    data = data.groupby('tradeDate').sum()
    data.plot(y=['netInflowS', 'netInflowM', 'netInflowL', 'netInflowXl'], kind='bar', figsize=(20, 8))
    
    return data

data = money_flow()
```

`板块涨跌情况`

```python
def industry_change():
    """行业涨跌平均涨幅
    """
    trade_days = lib.trade_pre_n_days()
    data = DataAPI.MktEqudGet(tradeDate=trade_days[-1], secID=set_universe('A'), 
                              field=['secID', 'ticker', 'preClosePrice', 'closePrice', 'tradeDate'])
    if data.empty:
        data = DataAPI.MktEqudGet(tradeDate=trade_days[-2], secID=set_universe('A'), 
                                  field=['secID', 'ticker', 'preClosePrice', 'closePrice', 'tradeDate'])
    
    data['up_ratio'] = data['closePrice'] - data['preClosePrice']
    data['is_up'] = [1 if i>=0 else 0 for i in data.up_ratio]
    data['is_down'] = [1 if i<0 else 0 for i in data.up_ratio]
    
    industry_ticker = DataAPI.EquIndustryGet(secID=set_universe('A'),
                                             field=['ticker', 'secID', 'secShortName', 'industryName2', 'industryName1','isNew', 'industry'])
    industry_ticker = industry_ticker[(industry_ticker.isNew == 1) & (industry_ticker.industry == '申万行业分类')]
    
    data = data.merge(industry_ticker[['ticker', 'secShortName', 'industryName2', 'industryName1']], on=['ticker'])
    
    industry2 = data.groupby('industryName2')
    industry1 = data.groupby('industryName1')
    
    ### plot avg ratio 
    fig_1 = industry1.mean()
    fig_1 = fig_1.sort(columns=['up_ratio'])
    ax = fig_1.plot(y=['up_ratio'], kind='bar', figsize=(20, 6), title=u'申万一级行业平均涨幅')
    for label in ax.get_xticklabels() + [ax.title]: 
        label.set_fontproperties(font) 
    
    fig_2 = industry2.mean()
    fig_2 = fig_2.sort(columns=['up_ratio'])
    ax = fig_2.plot(y=['up_ratio'], kind='bar', figsize=(20, 6), title=u'申万二级行业平均涨幅')
    for label in ax.get_xticklabels() + [ax.title]: 
        label.set_fontproperties(font) 
    
    ### plot up and down 
    fig_3 = industry1.sum()
    fig_4 = industry2.sum()
    ax = fig_3.plot(y=['is_up', 'is_down'], kind='bar', figsize=(20, 6), stacked=True, title=u'申万一级行业涨跌情况')
    for label in ax.get_xticklabels() + [ax.title]: 
        label.set_fontproperties(font) 
        
    ax = fig_4.plot(y=['is_up', 'is_down'], kind='bar', figsize=(20, 6), stacked=True, title=u'申万二级行业涨跌情况')
    for label in ax.get_xticklabels() + [ax.title]: 
        label.set_fontproperties(font) 
        
    return data

industry_data = industry_change()
```

`相关 lib 代码`

```python
from functools import wraps
import datetime as dt

############################################
### decorators
############################################

def wrapper_func_cache(func):
    @wraps(func)
    def wrapped(*args, **kwargs):
        if not hasattr(func, '__cache_result'):
            func.__cache_result = {}
        cache_key = hash((tuple(args), tuple(kwargs.iteritems())))
        if cache_key not in func.__cache_result:
            func.__cache_result[cache_key] = func(*args, **kwargs)
        return func.__cache_result[cache_key]
    return wrapped

############################################
### constants
############################################

@wrapper_func_cache
def const_trade_calendar():
    data = DataAPI.TradeCalGet(exchangeCD=u"XSHG,XSHE", beginDate=u"20050101", endDate=u"", field=['calendarDate', 'isOpen'])
    data = data[data.isOpen == 1]

    return list(data.calendarDate)

def trade_range(tstart, tend):
    """get all the trade days between tstart and ted
    """
    data = DataAPI.TradeCalGet(exchangeCD=u"XSHG,XSHE", beginDate=tstart, endDate=tend, field=['calendarDate', 'isOpen'])
    
    return list(data[data.isOpen == 1]['calendarDate'])
    
def trade_pre_n_days(tend='', n=1):
    """获取tend即当前n天的交易日
    """
    trade_days = const_trade_calendar() 
    tend = tend if tend else dt.datetime.today().strftime("%Y-%m-%d")
    
    index = trade_days.index(tend)    
    return trade_days[index - 30*n : index]


@wrapper_func_cache
def const_pre_trade_day(day):
    trade_days = const_trade_calendar() 
    index = trade_days.index(day)
    
    return trade_days[index - 1]
```
