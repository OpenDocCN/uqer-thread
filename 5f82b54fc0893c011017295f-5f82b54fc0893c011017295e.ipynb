基金选择范围为主动管理型股票基金，包括普通股票型、偏股混合型、均衡配置型、灵活配置型等。

```python
# 筛选策略最多需要之前多少年的数据
prev_year_n = 6

# 选择在prev_year_n年前成立，且已完成建仓（建仓期按3个月）的基金
est_year = cur_year - prev_year_n - 1
est_date = str(est_year) + '0930'
```

```python
def screen_active_funds(est_date):
    """筛选成立日期早于est_date的主动型基金，包括股票型、混合型。
    :returns: list，基金的secID。LOF基金返回后缀为XSHG、XSHE的secID。
    """

    # 选取混合型基金
    f_df = DataAPI.FundGet(
        category="H",
        operationMode="O",
        endDate=est_date,
        pandas="1")
    
    # 加入非股票型基金
    f_df = f_df.append(
        DataAPI.FundGet(
            category="E",
            operationMode="O",
            endDate=est_date,
            pandas="1"),
        ignore_index=True
    )
    
    # 进一步筛选
    filtered = []
    for i, f_info in f_df.iterrows():
        # 筛除指数基金
        if (f_info['indexFund'] == 'I') or (f_info['indexFund'] == 'EI'):
            continue
        # 筛除ETF基金
        if (f_info['etfLof'] == 'ETF'):
            continue
        # 筛除secID后缀为OFCN的LOF基金
        # f_df中每支LOF基金会对应两行，其中一个secID后缀为OFCN（其listStatusCd为'UN'），
        # 另一行后缀为XSHG、XSHE（这样的secID才能获取到LOF基金的净值数据)。
        if (f_info['etfLof'] == 'LOF') and (f_info['secID'][7:11] == 'OFCN'):
            continue
        # 筛除QDII、分级基金、保本基金
        if f_info['isQdii'] or f_info['isClass'] or f_info['isGuarFund']:
            continue
        # 筛除债券基金
        if '债' in f_info['secShortName']:
            continue
        
        filtered.append(f_info['secID'])

    return filtered
# end def
```

# 三、根据前N年年度涨幅排名初步筛选

## 计算基金的年度涨幅

```python
def calc_yearly_return(f_tk_list, year_list):
    """计算基金的年度涨幅。
    :param f_tk_list: list of ticker，基金列表。
                      部分基金的secID中途可能改变，如166006.XSHE在2015年2月变成了166006.XSHE2，
                      因此这里需要用ticker。
    :param year_list: list of int，计算年度涨幅的年份(如为当前年则计算年初到当前的涨幅）。
    """
    result_df = pd.DataFrame(columns=['ticker']).set_index('ticker')

    for f_tk in f_tk_list:

        for y in year_list:
            
            # 前一年年末日期
            start_date = str(y-1)+'1231'
            start_date = cal.adjustDate(start_date, BizDayConvention.Preceding)
            if start_date > cal.adjustDate(datetime.datetime.now(), BizDayConvention.Preceding):  
                # 前一年年末日期已大于当前日期，跳过
                continue
            # 该年度年末日期
            end_date = str(y)+'1231'
            end_date = cal.adjustDate(end_date, BizDayConvention.Preceding)
            if end_date > cal.adjustDate(datetime.datetime.now(), BizDayConvention.Preceding): 
                # 年末日期大于当前日期，计算到当前日期的涨幅
                end_date = cal.adjustDate(datetime.datetime.now(), BizDayConvention.Preceding)

            # 取得上一年年末和该年度年末基金复权净值
            df = DataAPI.FundNavGet(
                ticker=f_tk, 
                dataDate=start_date, 
                pandas="1")
            try:
                start_adj_v = df.loc[0, 'ADJUST_NAV']
            except KeyError:
                # 没取到值，跳过该年度
                start_adj_v = None

            df = DataAPI.FundNavGet(
                ticker=f_tk, 
                dataDate=end_date, 
                pandas="1")
            try:
                end_adj_v = df.loc[0, 'ADJUST_NAV']
            except KeyError:
                # 没取到值，跳过该年度
                end_adj_v = None

            # 计算年度涨幅
            if start_adj_v is not None and end_adj_v is not None:
                result_df.loc[f_tk, str(y)] = (end_adj_v - start_adj_v) / start_adj_v
            else:
                result_df.loc[f_tk, str(y)] = None
    return result_df
# end def
```

## 筛选前N年每年的年度涨幅排名在前X%的基金

```python
def select_top_yearly(yr_df, year, top=0.5):
    """筛选某年年度涨幅排名在top分位之前的基金。
    :param yr_df: DataFrame，index为基金ticker，须包含年度涨幅列，列名为'2012'、'2013'、'2014'……
    :param year: int，按该年年度涨幅排名排序。
    :returns: list，筛选出的基金ticker。
    """
    
    # top分位对应的序号
    top_idx = int(len(yr_df) * top)

    # 选取本年度涨幅排名前top_idx位的基金
    sorted_df = yr_df.sort_values(by=str(year), ascending=False)
    return sorted_df.iloc[:top_idx].index.tolist()
# end def
```

```python
def select_top(yr_df, cur_year, top=0.5, prev_n=6):
    """筛选cur_year之前prev_n年每年的年度涨幅排名在top分位之前的基金。
    :param yr_df: DataFrame，index为基金ticker，须包含年度涨幅列，列名为'2012'、'2013'、'2014'……
    :param cur_year: int，计算该年度之前的prev_n年（不包括该年）。
    :returns: list，筛选出的基金ticker。
    """
    
    # 前N年
    prev_years = [(cur_year - i - 1) for i in range(prev_n)]
    
    # 逐年筛选
    selected = None
    for y in prev_years:
        # 选取该年度涨幅排名前top分位的基金
        top_set = set(select_top_yearly(yr_df, y, top))
        # 各年度选出的基金取交集
        if selected is None:
            # 第一次，直接加入
            selected = top_set
        else:
            selected &= top_set
            
    return list(selected)
# end def
```

```python
# 筛选基金
sel_f_list = select_top(yearly_df, cur_year, top=0.5, prev_n=6)
```

## 计算策略选出的基金组合的年度平均收益

```python
def calc_group_return(g_tk_list, y_r_df, year):
    """计算基金组合的年度平均收益。
    :param g_tk_list: list of tickers，基金组合。
    :param y_r_df: DataFrame，index为ticker，包含年度涨幅。
    :param year: 计算该年度平均收益。
    :returns: %，基金组合该年度平均收益。
    """
    
    return y_r_df.loc[g_tk_list][str(year)].mean()
# end def
```

## 获取基金业绩评价指标

```python
# 基金业绩评价指标：
# 业绩持续性：HURST
# 择时能力：timingTM timingCL timingHM
# 选股能力：selectionTM selectionCL selectionHM
# 风险：maxDD secMDD thirdMDD annTotRisk
# 收益：annTotReturn
# 风险收益：sharpeRatio TREYNOR SORTINO

# 待获取的指标
# perf_ind_list = [
#     'annTotReturn', 'annTotRisk',
#     'maxDD', 'secMDD', 'thirdMDD',
#     'realizedAlpha', 'realizedBeta',
#     'sharpeRatio', 'TREYNOR', 'SORTINO', 'CALMAR',
#     'timingTM', 'timingCL', 'timingHM',
#     'selectionTM', 'selectionCL', 'selectionHM',
#     'HURST',
# ]

# 指标计算截至日期：上一年12月31日
indic_end_date = str(cur_year - 1) + '1231'
# 调整到最近的交易日（非交易日DataAPI.FundPerfIndicGet返回空表）
indic_end_date = cal.adjustDate(indic_end_date, BizDayConvention.Preceding)

# 指标时间窗口长度,1-近一个月;2-近三个月;3-近半年;4-年初至今;5-近一年;6-近两年;7-近三年;8-成立以来
window = 7

# 获取指标数据
perf_df = DataAPI.FundPerfIndicGet(
    dataDate=indic_end_date,
    ticker=sel_f_list,
    window=window,
    # field=perf_ind_list,
    pandas="1"
)

# 另一种方式
# perf_df = fa.performance(
#     symbol, start, end, benchmark=benchmark, 
#     # field=fields
# )
```
