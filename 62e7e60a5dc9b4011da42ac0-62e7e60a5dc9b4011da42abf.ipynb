# RPS个股趋势策略

1. RPS的计算逻辑：

RPSN（N表示N日前）
涨跌幅: EXTRS= （收盘价-N日前的收盘价）/ N日前的收盘价
然后按股票池中同一阶段的涨跌幅的数据，进行标准化的相对排名0~100
最后选出在相对排名的都表现优异的数据

2. 该案例选择54433法则：
	1年内涨跌幅排名 前50%， 6月内涨跌幅排名 前40%， 3月内涨跌幅排名 前40%， 1月内涨跌幅排名 前30%， 1周内涨跌幅排名 前30%
    选择沪深股通的标的作为股票池

明显没有选出符合标准的数据结果，因此需要对原始的涨跌幅数据进行去极值处理,下面对其去除极值处理

```python
def RPS_AfterWinsorize_Compute(stock_list):
    df1=DataAPI.MktEquPerfGet(adjType=u"1",secID=stock_list,ticker=u"", field=u"secID,chgPct1Y,chgPct6M,chgPct3M,chgPct1M,chgPct1W",pandas="1").set_index('secID')
    after_winsorize_1Y=winsorize(df1['chgPct1Y'].to_dict())
    df1.loc[after_winsorize_1Y.keys(), 'chgPct1Y_after']=after_winsorize_1Y.values()
    after_winsorize_6M=winsorize(df1['chgPct6M'].to_dict())
    df1.loc[after_winsorize_6M.keys(), 'chgPct6M_after']=after_winsorize_6M.values()
    after_winsorize_3M=winsorize(df1['chgPct3M'].to_dict())
    df1.loc[after_winsorize_3M.keys(), 'chgPct3M_after']=after_winsorize_3M.values()
    after_winsorize_1M=winsorize(df1['chgPct1M'].to_dict())
    df1.loc[after_winsorize_1M.keys(), 'chgPct1M_after']=after_winsorize_1M.values()
    after_winsorize_1W=winsorize(df1['chgPct1W'].to_dict())
    df1.loc[after_winsorize_1W.keys(), 'chgPct1W_after']=after_winsorize_1W.values()
    df1['1Y_Norm']=(df1['chgPct1Y_after']-df1['chgPct1Y_after'].min())/(df1['chgPct1Y_after'].max()-df1['chgPct1Y_after'].min())*100
    df1['6M_Norm']=(df1['chgPct6M_after']-df1['chgPct6M_after'].min())/(df1['chgPct6M_after'].max()-df1['chgPct6M_after'].min())*100
    df1['3M_Norm']=(df1['chgPct3M_after']-df1['chgPct3M_after'].min())/(df1['chgPct3M_after'].max()-df1['chgPct3M_after'].min())*100
    df1['1M_Norm']=(df1['chgPct1M_after']-df1['chgPct1M_after'].min())/(df1['chgPct1M_after'].max()-df1['chgPct1M_after'].min())*100
    df1['1W_Norm']=(df1['chgPct1W_after']-df1['chgPct1W_after'].min())/(df1['chgPct1W_after'].max()-df1['chgPct1W_after'].min())*100
    df_StockList=df1[(df1['1Y_Norm']>50)& (df1['6M_Norm']>60) & (df1['3M_Norm']>60)& (df1['1M_Norm']>70)& (df1['1W_Norm']>70)].index.tolist()
    return df_StockList

final_stockList = RPS_AfterWinsorize_Compute(stock_list)
print(final_stockList)
print len(final_stockList)
```
