-----------------

### 代表指数
[S&P 500 Low Volatility Index](https://us.spindices.com/indices/strategy/sp-500-low-volatility-index)

### 目标指数
HS300

### 选股
计算目标指数股票池中样本股过去100个交易日中的历史波动率，并挑选其中波动率最低的50只股票作为指数的成分股

### 加权
与传统指数市值加权不同，本指数根据股票波动率倒数为个股权重
