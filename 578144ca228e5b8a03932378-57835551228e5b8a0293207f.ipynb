### 5日调仓的问题
因为每5日调仓，可能会出现买入日期不同造成的误差，所以我把之后几天的也测试一下，看看会不会是因为买入日期造成的巧合。
（代码一样，日期改一下，就不贴出了）
经过测试看出，买入日期有一定的影响，
表现最好的测试，alpha达到78.6%，回撤也只有15.6%。
表现最差的测试，alpha达到50.7%，回撤29.3%
还记得之前的二八轮动模型，alpha只有39.4%，回撤16.4%
可以看出，SH50和创业板的alpha更高，但是回撤也变大了。

### 会不会是赶上创业板的大涨
问题又来了，观看持仓记录，发现这样的问题，大部分的盈利都是创业板的拉升。
继续尝试，把SH50的买入取消，只判断CYB
我选择上一个模型中表现最好的2013年5月9日开始的模型，之前的模型alpha78.6%，回撤15.6%。



测试完发现，如果只有CYB，回撤变小了。
测试了其他几个日期，也发现alpha降低，但是回撤变小。
