```python
import numpy as np
import pandas as pd
from CAL.PyCAL import *
import matplotlib.pyplot as plt
from datetime import datetime, timedelta
import matplotlib as mpl

stk = u"002572"

aa = DataAPI.MktStockFactorsDateRangeGet(secID=u"",ticker=stk,beginDate=u"20000101",endDate=u"",field=u"tradeDate,NetProfitGrowRate,NPParentCompanyGrowRate",pandas="1")
aa.columns = [u"日期",u"净利润增长率",u"归属母公司股东净利增长率"]
bb = aa.dropna()
bb = bb.set_index(u"日期")

fig = plt.figure(figsize=(15, 5))
ax = fig.add_subplot(111)
ax = bb.plot(ax=ax)

labels = [label.decode("utf-8") for label in bb.index.values]
ax.set_xticklabels(labels, fontproperties=font, fontsize=15)
ax.legend(prop=font)
ax.set_title(stk + u'历史净利增长率', fontproperties=font, fontsize=15)
ax.set_xlabel(u'日期', fontproperties=font, fontsize=15)
```
