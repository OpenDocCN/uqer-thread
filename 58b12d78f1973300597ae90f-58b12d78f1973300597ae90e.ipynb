```python
data1 = DataAPI.EquIndustryGet(industryVersionCD=u"010303",industry=u"",secID=u"",ticker=u"",intoDate=u"",field=u"secID,outDate,industryID1,industryName1",pandas="1")
print data1.head()
data3 = data1.groupby(['secID'])[['outDate']].max() 
print data3.head()
```

上面显示的内容是不是感觉没啥区别？仔细看……细看……看……嗯………………其实……最后显示的数据类型不同，这里要说明groupby对象很重要的特性，可以当做**dataframe来使**（比如groupby对象['X']或者groupby对象[['X']]）

```python
data3.reset_index(inplace=True)#重置索引
data = pd.merge(data1,data3,on=['secID','outDate'])#并merge在一起
print data.head()
```

准备一下行业列表

# 聚合运算

- 聚合还有哪些好用的函数？就像前面看到的.max()和.count()，这里有些常用的聚合运算函数，都是用来做数据统计的

```python
data.groupby(['industryID1','industryName1']).first()#取第一个非N值
```

为了避免内容过于冗长，列一下函数名（大伙自己把玩吧）：1）count() 非NA的个数；2）sum() 非NA和；3）mean() 非NA平均值；4）median() 非NA中位数；5）std()/var() 标准差/方差（统计里的两位好兄弟……）;6）min()/max() 最小/最大；7）first()/last() 第一/最后一个； 

- 这些压根满足不了我……我想DIY！D……I……Y你个头！（吐血）

```python
def garbage(x):
    return x.count()

data.groupby(['industryID1','industryName1'])['secID'].agg(garbage)
```

额……换了个姿势又显示了一遍count函数，但这里用agg来使用自定义函数，达到了DIY的效果……这里使用了**一般的分析三板斧**：1）分组groupby；2）['secID']筛选需要的列；3）agg/aggregate引用函数；

```python
data.groupby(['industryID1','industryName1'])['secID','outDate'].agg([('name_count','count'),('name_max','max')])
```

最后啰嗦一句，agg里的函数可以有两种形式：1）**[(名字1，函数1)，……，(名字n，函数n)]**（如上所示，每一列分别应用函数后返回名字以便区分，另外为什么函数能用 '' 引用？我也是一脸懵逼!?）；2）**{列名1：函数1，……，列名n：函数n}**(如下所示，这里对不同的列应用不同的函数，再加上迭代的方法更加灵活快捷）；PS:主要还是懒……所以搞出那么多"歪门邪道"

```python
data.groupby(['industryID1','industryName1'],as_index=False).agg({'secID':'count','outDate':'max'})
```

不知道各位有没有注意到上面的as_index，这个函数不会聚合索引，处女座的朋友们不要错过
