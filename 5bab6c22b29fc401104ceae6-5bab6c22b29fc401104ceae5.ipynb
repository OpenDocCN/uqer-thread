# look 1 look 因子叠加的有效性

## <font color=#0099ff>**多因子模型有效吗？**</font>
---
传统多因子模型：把收集到的基本面因子、量价因子做个大杂烩，根据得出的信号进行交易，
问题是：**因子叠加是否有效？**

- 看法1.一个因子能够选出一批股票，另外一个因子选出一批股票，两个因子发挥自己的特长，这个因子认为好的股票，那个因子也认为好的股票，集合的股票肯定是最好的股票

- 看法2.多因子模型会选出来一批中庸的股票。这个模型认为最好的股票，另一个模型里面认为是很差的，会造成这两个因子叠加后，选出来的股票都是这两个模型里面比较中庸的股票。两个因子打分后，在综合模型里面反而不是最好的股票了

#### `对基本面因子，量价因子做双分组分析`

**先用基本面因子把股票分成10组，然后根据价量模型再分10组，分析组合的收益情况**

- 如果因子的叠加是有效的：在热力图中有明显的线性关系，在基本面因子里面认为最好的股票、价量里面也是最好的股票。而最差的同理

- 如果多因子模型失去筛选能力：在热力图中不能找到明显的线性关系

#### `分析基本面因子与量价因子的相关系数`
**分析基本面与量价的全局相关性 & 分析基本面前10%与量价的相关性**

- 理想状态是基本面因子与量价因子不相关，这样多因子组合能取得最大增益

- 但是如果一定要从相关性角度考虑的话，则希望是正相关而不是负相关。我们希望基本面认为好的股票里面，价量因子也认为相对不错。

- 如果基本面认为好的股票，价量却认为是最差的，问题就比较大了。到时候选出来的肯定是中庸的股票，所以一般情况下我们希望两个模型有正相关关系是比较好的。

```python
import numpy as np
import pandas as pd
from datetime import datetime, timedelta
import time
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as st
import statsmodels.api as sm
from CAL.PyCAL import *
```

```python
def get_factor(trade_date_list, universe, factor_name):
    """
    Args:
        trade_date_list: 时间区间
        universe: 股票池
        factor_name: 因子名称
    Returns: 
        DataFrame: 因子数据(去除90天新股, st股)
    """
    factor = pd.DataFrame()
    for dt in trade_date_list:
        df_ST = DataAPI.SecSTGet(secID=universe, beginDate=dt, endDate=dt, field=['secID'])
        univ_ = [stk for stk in universe if stk not in list(df_ST['secID'])]
        dt_df = DataAPI.MktStockFactorsOneDayGet(tradeDate=dt, secID=univ_, field=['tradeDate', 'secID']+[factor_name])
        factor = factor.append(dt_df)
    factor = factor.pivot(index='tradeDate', columns='secID', values=factor_name)
    equ_info = DataAPI.EquGet(equTypeCD=u"A",secID=u"",ticker=u"",listStatusCD=u"",field=u"",pandas="1")
    equ_info = equ_info[['secID', 'listDate', 'delistDate']].set_index('secID')
    equ_info['delistDate'] = [x if type(x)==str else trade_date_list[-1] for x in equ_info['delistDate']]
    equ_info['delistDate'] = pd.to_datetime(equ_info['delistDate'], format='%Y-%m-%d')
    equ_info['delistDate'] = [x + timedelta(1) for x in equ_info['delistDate']]    
    equ_info['listDate'] = pd.to_datetime(equ_info['listDate'], format='%Y-%m-%d')
    equ_info['listDate'] = [x + timedelta(90) for x in equ_info['listDate']]
    factor.index = pd.to_datetime(factor.index)
    for sec in factor.columns:
        sec_info = equ_info.ix[sec]
        factor.loc[:sec_info['listDate'], sec] = np.NaN
        factor.loc[sec_info['delistDate']:, sec] = np.NaN     
    factor.index = [date.strftime('%Y-%m-%d') for date in factor.index]
    factor.index.name = 'tradeDate'
    factor.columns.name = 'secID'
    return factor

def pretreat_factor(factor_df, neu=True):
    """
    Args:
        factor_df: 因子数据
        neu: 是否中性化
    Returns:    
        DataFrame: 因子去极值，中性化（市值，行业风险），标准化的数据
    """
    pretreat_data = factor_df.copy(deep=True)
    for dt in pretreat_data.index:
        try:
            factor_dt = pretreat_data.loc[dt].dropna()
            factor_dt_dict = factor_dt.to_dict()
            if neu:
                pretreat_data.loc[dt] = pd.Series(standardize(neutralize(winsorize(factor_df.loc[dt].to_dict()), target_date=''.join(dt.split('-')), exclude_style_list=['BETA', 'RESVOL', 'MOMENTUM', 'EARNYILD', 'BTOP', 'GROWTH', 'LIQUIDTY', 'LEVERAGE'])))
            else:
                pretreat_data.loc[dt] = pd.Series(standardize(winsorize(factor_dt_dict)))
        except Exception as excp:
            print dt
            print excp
            continue
    return pretreat_data

def fac_q(factor_df, q1, q2):
    """
    Args:
        factor_df: 因子数据
        q1,2: 分位数
    Returns:    
        DataFrame: 因子N分位的数据
    """
    fac_q = factor_df.copy(deep=True)
    for dt in fac_q.index:
        tmp_ = fac_q.loc[dt].dropna()
        down = tmp_.quantile(q1)
        up = tmp_.quantile(q2)
        fac_q.loc[dt] = tmp_[(tmp_ <= up) & (tmp_ >= down)]
    return fac_q
```

```python
def plot_IC(fac1, fac2, title):
    """
    Args:
        fac1,2: 因子数据
        title: 标题
    Returns:    
        factor_IC
    """
    ic_data = pd.DataFrame(index=fac1.index, columns=['IC','pValue'])
    for dt in ic_data.index:
        tmp_1 = fac1.ix[dt]
        tmp_2 = fac2.ix[dt]
        cor = pd.DataFrame(tmp_1)
        cor2 = pd.DataFrame(tmp_2)
        cor.columns = ['corr']
        cor2.columns = ['cor2']
        cor['cor2'] = cor2['cor2']
        cor = cor[~np.isnan(cor['corr'])][~np.isnan(cor['cor2'])]
        if len(cor) < 5:
            continue
        ic, p_value = st.spearmanr(cor['corr'],cor['cor2'])   
        ic_data['IC'][dt] = ic
        ic_data['pValue'][dt] = p_value
    print '====================================='
    print title
    print 'IC: ', np.round(ic_data['IC'].mean(), 3)
    print 'P-value: ', np.round(ic_data['pValue'].mean(), 3)
    fig = plt.figure(figsize=(12, 4))
    ax1 = fig.add_subplot(111)
    ic_data.index = pd.to_datetime(ic_data.index)
    ax1.plot(ic_data.index, ic_data.IC, '#808080', label='IC')
    ax1.plot(pd.rolling_mean(ic_data.IC, window=3), '#FFA500', label='IC_mean(3 month)')
    ax1.legend(loc=1, fontsize=15)
    ax1.set_title(title, fontproperties=font, fontsize=15)
    ax1.grid(True)
```

```python
def plot_(start_date, end_date, univ_):
    """
    Args:
        start_date: 时间区间起点
        end_date: 时间区间终点
        univ_: 股票池
    Returns: 
        热力图：基本面和量价因子组合的下月回报
    """
    universe = DynamicUniverse(univ_)
    # 获取月末list
    cal_dates = DataAPI.TradeCalGet(exchangeCD=u"XSHG", beginDate=start_date, endDate=end_date).sort('calendarDate')
    cal_dates = cal_dates[cal_dates['isMonthEnd']==1]
    trade_month_list = cal_dates['calendarDate'].values.tolist()
    # 获取universe
    univ = []
    for date in trade_month_list:
        univ = list(set(univ + universe.preview(date, skip_halted=True)))
    # 合成基本面因子
    EP = 1 / get_factor(trade_month_list, univ, 'PE')
    EP = pretreat_factor(EP, neu=True)
    ROE = get_factor(trade_month_list, univ, 'ROE')
    ROE = pretreat_factor(ROE, neu=True)
    BP = 1 / get_factor(trade_month_list, univ, 'PB')
    BP = pretreat_factor(BP, neu=True)
    fundamentals_ = (EP + BP + ROE) /3
    fundamentals = fundamentals_.stack(dropna=False)
    fundamentals.name = 'fundamentals'
    # 合成量价因子
    REVS20 = get_factor(trade_month_list, univ, 'REVS20')
    REVS20 = pretreat_factor(REVS20, neu=True)
    HSIGMA = get_factor(trade_month_list, univ, 'HSIGMA')
    HSIGMA = pretreat_factor(HSIGMA, neu=True)
    VOL20 = get_factor(trade_month_list, univ, 'VOL20')
    VOL20 = pretreat_factor(VOL20, neu=True)
    volume_price_ = (REVS20 + HSIGMA + VOL20) /3
    volume_price = volume_price_.stack(dropna=False)
    volume_price.name = 'volume_price'
    # 合并
    cum = pd.concat([fundamentals, volume_price], axis=1)
    cum_ = cum.reset_index(drop=False)
    # 获得月收益率
    month_return = DataAPI.MktEqumAdjGet(secID=univ, beginDate=start_date, endDate=end_date, field=u"secID,endDate,chgPct,",pandas="1")
    month_return.rename(columns={'endDate': 'tradeDate', 'chgPct': 'month_return'}, inplace=True)
    month_return.sort_values(['secID', 'tradeDate'], inplace=True)
    month_return['next_month_return'] = month_return.groupby('secID')['month_return'].shift(-1)
    month_return.dropna(inplace=True)
    merge_df = cum_.merge(month_return, on=['secID', 'tradeDate'])
    # 设置10分位
    merge_df['fundamentals_q'] = merge_df.groupby(by='tradeDate')['fundamentals'].apply(lambda x:pd.qcut(x,10,labels=range(1,11)))
    merge_df['volume_price_q'] = merge_df.groupby(by=['tradeDate','fundamentals_q'])['volume_price'].apply(lambda x:pd.qcut(x,10,labels=range(1,11)))
    df_q = merge_df.groupby(by=['fundamentals_q','volume_price_q'])['next_month_return'].mean()
    df_q = df_q.reset_index(drop=False)
    df_q = df_q.pivot(index='fundamentals_q', columns='volume_price_q', values='next_month_return')
    # plot_heatmap
    plt.figure(figsize=(13, 10))
    pic = sns.heatmap(df_q, annot=True, cmap=sns.diverging_palette(220,10,as_cmap=True))
    # plot_corr
    plot_IC(fundamentals_, volume_price_, 'fundamentals corrwith volume_price')
    # plot_quantile_corr
    fundamentals_q = fac_q(fundamentals_, 0.95, 1.0)
    plot_IC(fundamentals_q, volume_price_, 'fundamentals 5% corrwith volume_price')
```

#### `2010-2016年的基本面与量价因子分析`
**因子双分组角度分析**
- 从热力图中看出，最蓝的区域（收益最低），出现在基本面和价量因子认为最差的股票组合。而最红的区域（收益最高），出现在基本面和价量因子认为最好的股票组合

**因子相关性角度分析**
- 从因子全局相关性的图中看出，（IC:-0.281， P-value:0.0）由于量价因子与基本面因子的方向相反，所以量价因子与基本面因子为正相关性，基本面认为好的股票，价量也认为是好的股票

- 从基本面前5%与量价的相关性的图中看出，（IC:-0.002，P-value:0.4）量价因子与基本面因子无相关性，这是最理想的状态

**我们在构建多因子组合时通常考虑前5%的因子，所以在2010-2016年基本面与量价因子叠加后，有增益的效果**

#### `2016-2018年的基本面与量价因子分析`
**因子双分组角度分析**
- 从热力图中看出，最绿的区域（收益最低），出现在基本面和价量因子认为最差的股票组合。然而我们无法在基本面和价量因子认为最好的区域找到最红的点，也无法发现明显的线性关系。

**因子相关性角度分析**
- 从因子全局相关性的图中看出，（IC:-0.27，P-value:0.0）由于量价因子与基本面因子的方向相反，所以量价因子与基本面因子为正相关性，基本面认为好的股票，价量也认为是好的股票

- 从基本面前5%与量价的相关性的图中看出，（IC:-0.048，P-value:0.408）量价因子与基本面因子有轻微正相关

#### `结论`
**从因子双分组的角度分析，从2016年年底以来，对于基本面因子认为好的股票，价量因子的区分能力却在下降**
- 这从一方面解释了：为什么2016-2018年传统的多因子模型失效了？ 如果多因子模型是做空而不是做多，多因子模型依然有效，从热力图中看出，基本面和量价因子认为最差的组合区域，收益还是最低，从做空角度看，2017-2018年多因子模型是有效的，但是从做多角度看却失效了，这或许就是原因。
