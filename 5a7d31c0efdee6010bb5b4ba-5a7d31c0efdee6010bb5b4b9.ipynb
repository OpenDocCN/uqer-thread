# 菲阿里四价法策略的改写和比较

介绍：昨天高点，昨天低点，昨天收盘，今天开盘，可并称为菲阿里四价，它是由日本期货冠军菲阿里实盘采用的主要突破交易的参照系，此外，因菲阿里主观心智交易的模式，决定了其在实际交易中，还大量结合应用了“阻溢线”的概念，即我们通常所说的压力、支撑线。
策略：趋势跟踪的一种，以昨日最高价作为上轨，最低价作为下轨，若当前价格向上突破上轨，多头开仓或加仓，若当前价格向下突破下轨，空头开仓或加仓，当日收盘清仓。本文以聚丙烯1801合约2017年下半年的回测数据为例，尝试了一些菲阿里四价法的改写，在初始条件相同的前提下，比较不同改写思路的效果。

思路一：Price向上突破上轨，多头开仓或加仓；Price向下突破下轨，空头开仓或加仓；尾盘清仓。

思路二：Price向上突破上轨或开盘价高于上轨，多头开仓或加仓；Price向下突破下轨或开盘价低于下轨，空头开仓或加仓；尾盘清仓。

思路三：Price向上突破上轨或开盘价高于上轨，多头开仓或加仓，Price向下突破上轨，多头平仓；Price向下突破上轨或开盘价低于下轨，空头开仓或加仓，Price向上突破下轨，空头平仓；尾盘清仓。

思路四：Price向上突破上轨或开盘价高于上轨，多头开仓或加仓，Price向下突破上轨，多头平仓；Price向下突破上轨或开盘价低于下轨，空头开仓或加仓，Price向上突破下轨，空头平仓；尾盘不清仓。

四种思路的年化收益率分别为65.7%、70.0%、53.4%和60.2%。从结果来看，思路二的收益最好，且收益波动相对稳定；由于思路三的策略在上下轨上的震荡导致反复平仓，致使收益下降；因为当日尾盘不执行清仓，导致思路四的收益波动较大，回撤明显。总结来说，在本例中菲阿里四价法的有效性还是挺高的，一旦突破上下轨则应相信趋势加仓，反复平仓会使收益降低，日间交易相对于日内交易风险较大，尾盘清仓更有利于降低风险。
