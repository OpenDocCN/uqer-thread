```python
import tushare as ts
import numpy as np

data = ts.get_hist_data('600848',start='2017-11-01',end='2019-12-01')
#print(data)
data['date'] = data.index
data['year'] = data['date'].apply(lambda x: x[0:4])
data['month'] = data['date'].apply(lambda x: x[5:7])
#data.index[0][5:7]
 
def MaxDrawdown(return_list):
    '''最大回撤率'''
    i = np.argmax((np.maximum.accumulate(return_list) - return_list) / np.maximum.accumulate(return_list)) 
    if i == 0:
        return 0
    j = np.argmax(return_list[:i])
    return (return_list[j] - return_list[i]) / (return_list[j])


maxdown = []
for year in data['year'].unique():
    for month in data['month'].unique():
        return_list = data[(data['year']==year)&(data['month']==month)]['close']
        if not len(return_list)==0:
            maxdown.append(MaxDrawdown(return_list))

print(max(maxdown))


import pandas as pd
data2 = pd.read_csv('/Users/adrienuchiha/Downloads/000001.csv',encoding = 'gbk')
#data2.info()
data2['year'] = data2['日期'].apply(lambda x: x[0:4])
data2['month'] = data2['日期'].apply(lambda x: x[5:7])
data2 = data2[(data2['year']>='2008')&(data2['year']<='2019')]
maxdown2 = []
time = []
for year in data2['year'].unique():
    for month in data2['month'].unique():
        return_list = data2[(data2['year']==year)&(data2['month']==month)]['收盘价']
        if not len(return_list)==0:
            maxdown2.append(MaxDrawdown(return_list))
            time.append(year+month)

time.reverse()
maxdown2.reverse()

import matplotlib.pyplot as plt
from matplotlib.pyplot import  MultipleLocator
xmajorLocator = MultipleLocator(6)
fig, ax = plt.subplots(figsize = (12,8),dpi =120)
ax.xaxis.set_major_locator(xmajorLocator)
plt.xticks(rotation=30)
plt.plot(time,maxdown2)
```
