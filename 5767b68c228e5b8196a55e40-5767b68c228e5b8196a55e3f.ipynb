15岁参加高考并进入复旦大学学习物理；16岁开始交易，至今已经有16年国内外交易经历；历经中国多轮牛熊市。沈天瑞博士的身上，有着许许多多的标签。如今，作为垒土投资首席投资官、创始合伙人，沈天瑞正引领着具有丰富本土市场经验的团队，搭载全球先进的系统性框架，开发出适合中国市场的对冲交易策略。一个新的标签，正在他的身上，熠熠发光。

## 兴趣是最好的老师 ##

`兴趣是最好的老师，实践是最好的课堂。`沈天瑞博士的量化之路，从个人兴趣开始。
 
沈天瑞自幼就按照自己的兴趣来学习，无论是数学还是计算机。他接触计算机的时间非常早，87、88年的时候，其就读的上海市实验学校就建立了学校机房。因此，从小学1年级开始，沈天瑞就接触到了计算机。无论是Basic、Fortran、还是Pascal，这些编程语言他都全部自学过，包括后来的C、C++。
 
但是当使用了多种计算机语言后，他发现`编程首先是一种思维，就是把自己所想的逻辑写出来，`接下来是算法，所以`计算机其实是一种工具`。当时沈天瑞的兴趣更多在自然科学上， 因此，大学时他并没有选择计算机专业，而是选择了物理(光源与照明)。到了硕士、博士阶段，沈天瑞开始着迷于探索人脑进行思维和成长的模式，因此选择了心理学。沈天瑞称，他所有专业的选择都是基于兴趣的驱动，而把自己感兴趣的事情分析透彻让他充满成就感。
 
开始做投资也是如此，尽管当时国内并没有Quant这个职业，但沈天瑞已经主动将编程、技术分析和投资结合了起来。读大学时，他就开始把自己的奖学金放进股票账户，然后从最基本的模式入手，寻找规律进行分析。市场一直在变，每天都有新情况，从瞬息万变中找到不变，让他深深沉迷。
 
他说：“`这其实是一种本能，当你对一个东西有兴趣的时候，你就会不断去想一些研究方法，把这个东西搞明白。兴趣是交易非常重要的起点。`而并不是说，我觉得Quant是一个很好的行业，所以我要努力进入这个行业。你可以先用比较简单的方法去尝试，然后在实践过程中，不断地学习，使自己的知识体系更加专业完善。”

## 跨界不是投身量化的阻碍 ##

非金融专业出身的沈天瑞投身金融行业，被很多人视作是一次跨界尝试。但在他的眼中，自己的专业知识非但不是量化事业的阻碍，反而对从事量化投资有着极大的帮助。比如，从价格的随机变动可以映射一些粒子物理方面的模型。
 
沈天瑞指出，就研究本身而言，传统意义上的研究都是自上而下的：先设立一个假说，然后通过实验设计用数据去验证假说正确与否，再把研究目标一步一步分解，做到精准。而现在，垒土投资则通过各种各样的对冲工具，对研究出来的结果中的一部分进行精准下注。从这个角度而言，研究方法论、统计学和数学知识都是必不可少的。
 
谈及心理学，沈天瑞介绍，自己所学的心理学并不是社会心理学，而是属于认知心理学，与量化投资其实有很多相通之处。人脑是如何思维的、对各类信息大脑哪些部位相关，反应呈现怎样的模式，这与市场针对某一信息、某一事件驱动做出反应的原理是一致的。大脑各个分区之间会有并行与相互影响，类似于投资者相互之间会有一些传导。从研究角度来说，这些都是相通的。“当我们想方法建模时，可以从这个点切入，也可以从那个点切入，但是无论从哪个点，可能都不是完备的。但在一定范围之内，可能它是规律的，那它就是有效的。”沈天瑞博士指出，如果分解出预测价格变动的有效规律，就可以放到自己的模型中。相较于普通研究，交易模型的结果直接影响盈亏，容不得有马虎的地方。
 
大道相通，不论是让普通人望而生畏的物理学还是如今火热的金融市场，`思考或研究是人的本能`，人的大脑本能地会去归纳一些东西。沈天瑞发现，尽管有经验的交易员会总结出一些自己的交易经验，但做研究一定要遵循一个系统的研究方法，这些方法论的知识，不管什么学科，都是会涉及到的。所以，关键不在于学的什么学科，而在于学习过程中的思维训练和严谨的态度，这些对最后的量化建模，都是非常有价值的。
 
`编程其实就是一个工具。`沈天瑞认为，对于编程的掌握程度最好要做到把它作为一种本能，进行量化投资时，效率就会得到极大提高。“量化投资对编程的要求其实就两方面，一是一定要尽量少bug，因为它需要绝对的精确。这点上只需要细心，对技术并没有太大要求。另一方面就是如何让脚本运行得更快，提高效率。”他指出，不管是回测或是最后生产程序，如果速度很慢，对量化的影响就很大，除了堆叠硬件，这点需要更具创造力。“很多知识其实我们在高中就已经学了，只是长时间不用慢慢就遗忘了，所以我们需要训练自己，保持开放的心态，不断学习。” 作为一个别人眼中的“跨界者”，沈天瑞在金融与其他专业的结合上有着自己的见解。“跨界有时可以跳出传统金融思想的束缚，例如行为金融。但是掌握基础金融知识是必须的，起码要知道自己在交易的是什么，否则很多对别人来说不是黑天鹅的事情，对自己来说就是。”

## 量化投资不论出身 ##

“前些年许多出国留学的理工科博士毕业后都会选择量化，这也可能是很多人发现做量化的人大多出身世界顶尖学校、且学历很高的主要原因。但这并不表明其他学生就不能从事量化投资。”沈天瑞博士指出，研究的一些基本技能和习惯是经过训练的人要比没有经过训练的好，但是创造力或研究的视角，这些却是很难说的。沈天瑞一直认为`中国本土学生其实功底都很好，也有自身的独特优势，对跟国内市场理解更深、更接地气，只需稍加培养，释放出的创造力将比机械、教条照搬海外的更加强大。`沈天瑞表示，垒土投资正是希望能中西结合，培养一支本土团队，帮助他们在量化市场大放异彩。
 
从系统框架和方法的层面来说，沈天瑞认为海外的机构确实比中国市场现有的水平领先很多，也更为精细。量化投资在国外发展相对成熟，很多方法理论沉淀已久，在国外的学习和工作经历的确可以加速人的成长。但是随着越来越多从事量化投资的海外回国人员把最先进的理论和思想带回国内，国内外在方法论上的差距正在逐步缩小，包括通联数据现在构建的这套研究框架系统也是之前BGI一直都在使用的。沈天瑞指出，`像优矿这样的平台，其实已经植入了一些国外先进的研究方法。在这样的环境下，国内已经能够培养和养成之前认为在海外领先的对冲基金公司才能训练出来的一些具有良好的投资习惯和素养的量化人才。`

## 沈天瑞眼中的量化投资 ##

为什么会选择投身量化投资？量化与传统交易的区别又在哪里？沈天瑞认为，量化投资与传统交易相比，最主要的区别在于：`量化投资是以一个更加理工科学的视角去看待市场的变化，然后试图去给市场进行建模、预测。`而传统的投资，比如自下而上的投资先分析该公司基本面，从而对这个公司进行价值定位，再评判是否存在定价偏差和潜在投资机会。量化从某种程度上来说，是撇开所有这些，`用纯粹数学和数据的视角去看待问题。`过去量化很依赖自上而下的研究，用数理统计的方法去验证假说，寻找规律。但再过十年，机器学习将是一个新方向。“人脑不可能像机器一样将所有的数据信息都完整地记录在案，并能随时快速提取。在如今这个信息膨胀的时代，资讯流动太快，已经远远超出人脑的可控范围。”沈天瑞博士表示，人作为一种生物，本身存在很多生理瓶颈，不管是注意力的持续性还是思维的局限性，这些瓶颈都是不可否认的。机器学习恰恰是能够帮助我们打破这些瓶颈的有效手段。将很多机器能够胜任的工作编成程序，交给机器去做，那么完成同样的事情会更高效，最终降低所有市场参与者的成本。而且机器学习不仅能够从大数据入手产生类似基本面投资价值发现的功能，从研究角度上也可能产生自下而上的路径，带来启发。

沈天瑞觉得作为一名Quant，有`3点核心特质`：

1、用系统化的方式决问题，把问题分解成一个个小模块，做好每一个模块，有累积性地不断精细化改进，结果是最后的输出。

2、对研究有兴趣。做研究，很有可能你做了100次尝试，最后只有两个有成果。这种情况下，只有你真的对研究感兴趣，才能沉得住气继续做研究。 正确的研究方法可以帮助少走很多弯路。

3、研究、诊断、改进的正反馈循环。量化模型基于历史数据，在样本外实盘运行时良好的正反馈循环是：严谨、透彻的研究，科学的监测/跟踪/诊断体系来接收市场交易结果的反馈，根据反馈进一步改进模型，模型更完善，同时也对模型更具信心。

## 工作即兴趣，兴趣驱动工作 ##

“其实老实说，我本人最大的兴趣爱好就是做研究，与研究有关的我也会比较关注。其次我是水瓶座，比较天马行空一点，对一些新奇的、比较酷的黑科技之类的我都会比较感兴趣。”沈天瑞如是说。
 
`交易有时会与兴趣爱好相关。`比如海外很多主动交易员都是运动背景出身，他们身上的一些特质，像进取心、对状态的控制能力等，实际上和交易是相通的。
 
谈到做量化这行的感受，沈天瑞坦言，压力是一定存在的，而他的解压方法是：兴趣驱动工作。他认为，压力取决于自己对研究的把握程度，这是一个循环。如果对自己的研究比较有信心，那么在跑研究时，其实就像是抚养孩子，看有哪些优缺点，想办法帮他成长。这是正循环，这种情况会给自己带来成就感。但还有另一种情况，研究过程中有疏忽，对自己的研究信心不足。在交易时，人对信息的接收是有选择性的，此时就会更容易接收到一些负面信息，然后想办法去改变，最后就会觉得市场压力不断堆积。所以，可能在一开始对研究多花些功夫，后面的过程就会轻松很多。
 
`“九尺高台，起于垒土；千里之行，始于足下。”`沈天瑞说，当初把这句话定为垒土投资的标语，正是因为他希望一步步去把一些小的事情做好。从研究的角度，也是希望团队能仔细认真做好每一个研究框架，把每一个模块做细做精，最后输出的业绩是根据某一些模块打造出来的结果：投资者有什么偏好，垒土就可以生成相应的结果。从这方面自下而上来说，他们并不是从一个结果去倒推如何开发策略，而是更倾向于最初就把模块做好，让它能自动生成不同的有效结果。

## 附录 ##

1、当问沈天瑞对刚刚进入我们这一行的包括优矿的用户、学生、以及量化投资爱好者有什么推荐的学习资料及网站时，他第一推荐的就是BGI大牛出的《主动投资组合管理》。
 
沈天瑞说这是非常经典的一部著作，里面大部分其实没有涉及太深的知识，都是一些金融知识，对理工科的学生来说，其实没有太大的门槛。对于要做市场中性或者要做阿尔法以及要通过量化来做因子研究的投资者来说，基本上也是一个比较标准的教材。他认为这本书里有很多科普的知识来告诉学习者要用哪些基础的标准去衡量、判读，然后自己动手进行实践。在这个过程中，还可以在学术网站搜索一些经典文献，这对于量化入门都是有一定帮助的。

2、**沈天瑞博士寄语：**
其实大家都是在一个学习的过程当中，市场也一直在变，所以说保持一个较为开放的心态才是最重要的。技术更多的只是一个工具，如果热爱研究，技术上的鸿沟是很容易跨越的，但是要保持愿意不断学习研究的心态。
