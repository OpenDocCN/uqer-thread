# 股息率

```python
import pandas as pd
import numpy as np
import datetime
date=DataAPI.TradeCalGet(exchangeCD=u"XSHG",beginDate=u"20070101",endDate=u"20171231",field=u"calendarDate,isOpen",pandas="1")
date=date[date.isOpen==1]
date.index=range(len(date))  # 得到全部的交易日
pool=set_universe('A')   # 得到全部的股票池
div_df=DataAPI.MktAdjfGet(exDivDate=u"",secID=pool,beginDate=u"20070101",endDate=u"20171231",field=u"",pandas="1")
```

```python
from __future__ import division
# ---------------------计算发放股息后填权所需要的时间（剔除高送转）---------------------
# 得到除权的日期
div_df=div_df[np.isfinite(div_df['perCashDiv'])]   # 剔除非现金分红
div_df=div_df.fillna(0)
div_date=div_df.exDivDate
div_date=list(set(div_date))
div_date.sort()

result=[]
fail_to_fill=pd.DataFrame(columns=['ticker','ex_time'])
indx=0 # 记录没有填权的参数
# 找出该股票除权日的股价和前一个交易日的股价
for i in div_date:
    stocks=list(div_df[div_df.exDivDate==i].ticker)
    df=DataAPI.MktEqudGet(tradeDate=i,ticker=stocks,isOpen="",field=u"",pandas="1")
    
    # 找出用了几天填权
    for s in stocks:
        prices=DataAPI.MktEqudGet(ticker=s,beginDate=i,endDate=u"20171231",isOpen="",
                                  field=u"ticker,secShortName,tradeDate,highestPrice",pandas="1")
        preDivPrice=df[df.ticker==s].preClosePrice.iloc[0]    # 复权价
        rate_of_tran=div_df[(div_df.ticker==s) & (div_df.exDivDate==i)].perShareDivRatio + div_df[(div_df.ticker==s) & (div_df.exDivDate==i)].perShareTransRatio + 1
        rate_of_tran=rate_of_tran.iloc[0]
        if rate_of_tran==1:   # 如果转增比例等于1（因为在公式里面加了1），证明只有现金分红
            pass
        else:
            preDivPrice=preDivPrice / rate_of_tran  # 不考虑分红的除权价
        
        try:
            fill_date=prices.ix[prices[prices.highestPrice>=preDivPrice].index[0]].tradeDate
            fill_date=datetime.datetime.strptime(fill_date,'%Y-%m-%d')
            start_date=datetime.datetime.strptime(i,'%Y-%m-%d')
            days=(fill_date - start_date).days
            result.append(days)
        except:
            fail_to_fill.loc[indx,'ticker']=s
            fail_to_fill.loc[indx,'ex_time']=i
            indx+=1
            pass
        
result=np.array(result,dtype='int')
success_rate=len(result)/(len(result)+len(fail_to_fill))
print '总样本数：',len(result)+len(fail_to_fill)
print "填权率为：",success_rate*100," %"
print "平均数为：", np.average(result) 
print "中位数为：", np.median(result)
counts = np.bincount(result)
print "众数为：", np.argmax(counts)
print "方差为：", np.var(result)
print "最大天数为：",max(result)
print "最小天数为：",min(result)
```
