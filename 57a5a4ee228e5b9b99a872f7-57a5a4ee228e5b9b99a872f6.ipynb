### 6、一个多因子模型一般有多少个因子
---

 - 一般以有多少类因子来衡量模型涵盖的因子数量，同一类可能涵盖好几个因子，比如估值类因子可能就由市盈率PE、市净率PB组成；
 
 - 总体来说，一个模型可能涵盖10类左右的因子，至于最后因子个数的绝对值大概20个左右
