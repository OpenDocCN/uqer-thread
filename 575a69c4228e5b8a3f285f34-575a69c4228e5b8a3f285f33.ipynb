[在CCI策略--基础篇](https://uqer.datayes.com/community/share/57593768228e5b8a3d2861c0)中提到，CCI指标过于敏感，开仓平仓信号经常出现过早出现，为了解决这个问题在CCI的基础上衍生出CCI Correction策略。简单讲就是利用大周期或者周期参数较大的CCI值判断市场是在牛市还是熊市，利用小周期或者周期参数较小的CCI值判断买卖点。追求有些交易者提到的“确定的机会”。
在继续阅读下面文档和代码前，请阅读参考文献以了解CCI Correction的详细原理。[CCI Correction](http://stockcharts.com/school/doku.php?id=chart_school:trading_strategies:cci_correction)


以下策略示例以上证50参数为benchmark，以上证50ETF（510050）为交易标的。主要参数是CCI长、短周期的参数的配合，就粗浅的经验，长周期参数在短周期参数的3~5倍间，表现较好。比如示例中长周期参数为26，短周期参数为8.
CCI Correction的优点是只在市场走好时才开仓，很少出现抄底抄在半山腰上的情景。缺点主要有两个，一是放弃了超跌反弹行情；二是当超跌反弹行情出现时容易被博弈，当强势超跌反弹进入尾声，CCI Correction策略容易误判市场进入牛市，CCI Correction交易者开仓，而在在底部抄底的交易者趁机出货，超跌反弹行情结束，CCI Correction交易者被迫割肉离场。
