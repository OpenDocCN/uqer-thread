# 抱团研究之数据 - 基金重仓股
本文希望开始一段抱团现象的研究系列。作为第一篇，本章节着手利用优矿API筛选基金重仓股。

大体思路：
1） 利用公募基金的持股披露信息：季报中的前10大持股，半年报和年报中的全部持股信息
2） 加总所有基金对某股票的持有总市值，计算其与总流通市值的比值

### 问题一：信息披露时间不一致
基金季报 - 每个季度结束之日起十五个工作日内，披露前十大股票持仓 - 例如 2020四季报 披露于 1/21/2021
基金半年报 - 上半年结束之日起六十日内，披露全部股票持仓 - 例如 2020半年报 披露于 8/29/2020
基金年报 - 每年结束之日起九十日内，披露全部股票持仓 - 例如 2019年报 披露于 4/9/2020

### 解决方案
考虑到后续可能会设计季度频率的策略，这里保留两种选项：
1） QtrReport_only = True：假设只利用季报，可以在季度结束之后的第一个月底得到各基金前十大持股信息
2） QtrReport_only = False:利用半年报和年报中的全部持股信息，但设计策略时需注意报告发布时间的之后

```python
import datetime
import pandas as pd
import numpy as np
```

```python
# 假设取2020三季报的数据
report_dt = '2020-09-30'
QtrReport_only = True # 由于是三季报，所以实际上没有体现差异

# 取所有基金的股票持仓信息
df = DataAPI.FundHoldingsGet(reportDate=report_dt, secType="E",field=['secID','SecShortName','reportDate','holdingSecID','holdingTicker','marketValue'])

# 如有仅用季报信息，则只可见前10大持股，但信息于季度后一个月可得
if QtrReport_only == True:
    df['marketValueRank'] = df.groupby(['reportDate','secID'])['marketValue'].rank(ascending=False)
    df = df[df['marketValueRank']<=10] 
    
# 仅保留A股
df['exchange'] = df['holdingSecID'].apply(lambda x: x[-4:])
df = df[df['exchange'].isin(['XSHE','XSHG'])] 
df.head()
```

```python
# 计算各股票的基金持股总市值
df_holding_pos = df.groupby('holdingSecID')['marketValue'].sum().reset_index().rename(columns={'holdingSecID':'secID',
                                                                                               'marketValue':'fundHolding'})
# 取当季最后一个交易日的自由流动市值
trade_dt = '2020-09-30'
df_holding_pos = df_holding_pos.merge(DataAPI.MktEqudGet(secID=df_holding_pos['secID'],
                                                         tradeDate=trade_dt,field=['secID','negMarketValue','marketValue']),
                                      how='left', on='secID')
# 基金持股比率=基金持股市值/自由流动市值
df_holding_pos['fundHoldingRatio'] = df_holding_pos['fundHolding'] / df_holding_pos['negMarketValue']

# 剔除科创板
df_holding_pos = df_holding_pos.merge(DataAPI.EquGet(secID=df_holding_pos['secID'],
                                                     field=['secID','secShortName','ListSector','ListSectorCD']),
                                      how='left', on='secID')
df_holding_pos = df_holding_pos[df_holding_pos['ListSectorCD']!=4]

# 仅保留基金持股比率>=1%的股票，减少输出数据
df_holding_pos = df_holding_pos[df_holding_pos['fundHoldingRatio'] >= 0.01]

df_holding_pos.sort_values(by='fundHolding', ascending=False).head(10)
```

### 问题二：分级基金的重复统计
经分析，发现上述统计中，未剔除分级基金的重复计算。

```python
# 下面以“邮储银行”在2020三季报的数据为例进行展示，发现有分级基金的重复计算
df[df['holdingTicker']=='601658'].sort_values(by='marketValue', ascending=False).tail(15)
```

### 解决方案：
利用基金信息表对分级基金进行识别和去重

```python
# 取基金分级信息
df_fund_name = DataAPI.FundGet(field=['secID','secShortName','className']) 
df_fund_name['secShortName_nodup'] = df_fund_name.apply(lambda x: x['secShortName'] if pd.isnull(x['className']) else x['secShortName'][:-1-len(x['className'])], axis=1)

# 未去重
report_dt = '2020-09-30'
df = DataAPI.FundHoldingsGet(reportDate=report_dt, secType="E",field=['secID','SecShortName','reportDate','holdingSecID','holdingTicker','marketValue'])
print('2020-09-30 - 邮储银行 - 基金持股市值（未去重）- {:,.0f}'.format(df[df['holdingTicker']=='601658']['marketValue'].sum()))

# 利用分级信息去重
df_nodup = df.merge(df_fund_name, how='left', on='secID')
df_nodup = df_nodup.drop_duplicates(subset=['holdingSecID','reportDate','secShortName_nodup','marketValue'])
print('2020-09-30 - 邮储银行 - 基金持股市值（去重）- {:,.0f}'.format(df_nodup[df_nodup['holdingTicker']=='601658']['marketValue'].sum()))
```

### 取2005年来的基金重仓股信息

1) 编写函数

```python
def fundHoldingRatio_get(report_dt, trade_dt, df_fund_name, QtrReport_only = True):
    # 取所有基金的股票持仓信息
    df = DataAPI.FundHoldingsGet(reportDate=report_dt, secType="E",field=['secID','SecShortName','reportDate','holdingSecID','holdingTicker','marketValue'])

    # 如有仅用季报信息，则只可见前10大持股，但信息于季度后一个月可得
    if QtrReport_only == True:
        df['marketValueRank'] = df.groupby(['reportDate','secID'])['marketValue'].rank(ascending=False)
        df = df[df['marketValueRank']<=10] 

    # 仅保留A股
    df['exchange'] = df['holdingSecID'].apply(lambda x: x[-4:])
    df = df[df['exchange'].isin(['XSHE','XSHG'])] 

    # 利用分级信息剔除重复
    df_nodup = df.merge(df_fund_name, how='left', on='secID')
    df_nodup = df_nodup.drop_duplicates(subset=['holdingSecID','reportDate','secShortName_nodup','marketValue'])

    # 计算各股票的基金持股总市值
    df_holding_pos = df_nodup.groupby('holdingSecID')['marketValue'].sum().reset_index().rename(columns={'holdingSecID':'secID',
                                                                                                   'marketValue':'fundHolding'})
    # 取当季最后一个交易日的自由流动市值
    df_holding_pos = df_holding_pos.merge(DataAPI.MktEqudGet(secID=df_holding_pos['secID'],
                                                             tradeDate=trade_dt,field=['secID','negMarketValue','marketValue']),
                                          how='left', on='secID')
    # 基金持股比率=基金持股市值/自由流动市值
    df_holding_pos['fundHoldingRatio'] = df_holding_pos['fundHolding'] / df_holding_pos['negMarketValue']

    # 剔除科创板
    df_holding_pos = df_holding_pos.merge(DataAPI.EquGet(secID=df_holding_pos['secID'],
                                                         field=['secID','secShortName','ListSector','ListSectorCD']),
                                          how='left', on='secID')
    df_holding_pos = df_holding_pos[df_holding_pos['ListSectorCD']!=4]

    # 仅保留基金持股比率>=1%的股票，减少输出数据
    df_holding_pos = df_holding_pos[df_holding_pos['fundHoldingRatio'] >= 0.01]
    df_holding_pos = df_holding_pos.sort_values(by='fundHolding',ascending=False)
    df_holding_pos['reportDate'] = report_dt
    return df_holding_pos[['reportDate','secID','fundHoldingRatio','fundHolding','secShortName','ListSector']]
```

3) 取季报日期和季末最后交易日

```python
# 取季末交易日
trading_cal = DataAPI.TradeCalGet(exchangeCD=u"XSHG,XSHE",beginDate=u"20050101",endDate=datetime.datetime.today(),field=['calendarDate','isQuarterEnd','isOpen']).drop_duplicates()
trading_cal['QtrID'] = trading_cal['calendarDate'].apply(lambda x: x[:7])
trading_cal_qtr = trading_cal[trading_cal['isQuarterEnd']==1]

# 取季报报告日期
ME_dates = trading_cal.sort_values(by='calendarDate').drop_duplicates(subset=['QtrID'], keep='last')[['QtrID','calendarDate']].rename(columns={'calendarDate':'ME_report_date'})
trading_cal_qtr = trading_cal_qtr.merge(ME_dates, how='left', on='QtrID').rename(columns={'calendarDate':'ME_trade_date'}).set_index('QtrID')

# 保存为数据字典
dict_report_trade_dt = dict(zip(trading_cal_qtr['ME_report_date'], trading_cal_qtr['ME_trade_date']))
trading_cal_qtr.tail()
```

4) 遍历报告日期，循环选取保存

```python
df_ratio_total = pd.DataFrame()
for qtrID in trading_cal_qtr.index:
    report_dt, trade_dt = trading_cal_qtr.loc[qtrID, 'ME_report_date'], trading_cal_qtr.loc[qtrID, 'ME_trade_date']
    #print('Pulling Fund Holding Ratios for quarter {:s} - report date {:s} - trading date {:s}'.format(qtrID, report_dt, trade_dt))
    df_ratio_total = pd.concat([df_ratio_total, fundHoldingRatio_get(report_dt=report_dt, trade_dt=trade_dt, df_fund_name = df_fund_name, QtrReport_only = True)])
    
df_ratio_total[df_ratio_total['reportDate']=='2020-12-31'].head(10)
```
