```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.optimize as sco
import scipy.interpolate as sci
from scipy import stats

a=DataAPI.MktEqudAdjGet(secID=u"",ticker=u"000651",tradeDate=u"",beginDate=u"20140528",endDate=u"20170526",isOpen="",field=u"",pandas="1")
df=a[['tradeDate','closePrice']]
s000651 = df['closePrice']
s000651.name = '000651'

b=DataAPI.MktEqudAdjGet(secID=u"",ticker=u"600519",tradeDate=u"",beginDate=u"20140528",endDate=u"20170526",isOpen="",field=u"",pandas="1")
df1=b[['tradeDate','closePrice']]
s600519 = df1['closePrice']
s600519.name = '600519'

data=pd.DataFrame({'000651':s000651, '600519':s600519})
data=data.dropna()

log_returns = np.log(data / data.shift(1))        
log_returns.hist(bins=50, figsize=(12, 9))       
plt.show()   

rets = log_returns                               
year_ret = rets.mean() * 252                      
year_volatility = rets.cov() * 252               

number_of_assets =2                                           
weights = np.random.random(number_of_assets)                     
weights /= np.sum(weights)                                      

# 假设有2000组投资组合，每一组投资组合由一组随机权重组成，则可以这样产生2000组投资组合：
portfolio_returns = []
portfolio_volatilities = []
for p in range (2000):
    weights = np.random.random(number_of_assets)
    weights /= np.sum(weights)
    portfolio_returns.append(np.sum(rets.mean() * weights) * 252)
    portfolio_volatilities.append(np.sqrt(np.dot(weights.T,np.dot(rets.cov() * 252, weights))))
# 得到2000个组合收益率和波动率
portfolio_returns = np.array(portfolio_returns)
portfolio_volatilities = np.array(portfolio_volatilities)

# 接下来将2000个组合收益率和波动率用散点图描述，并且计算每个夏普率
plt.figure(figsize=(9, 5))
plt.scatter(portfolio_volatilities, portfolio_returns, c=portfolio_returns / portfolio_volatilities, marker='o')
plt.grid(True)
plt.xlabel('expected volatility')
plt.ylabel('expected return')
plt.colorbar(label='Sharpe ratio')
plt.show()

print ''
def statistics(weights):
    # 根据权重，计算资产组合收益率/波动率/夏普率。
    # 输入参数
    # ==========
    # weights : array-like 权重数组
    # 权重为股票组合中不同股票的权重
    # 返回值
    # =======
    # pret : float
    #      投资组合收益率
    # pvol : float
    #      投资组合波动率
    # pret / pvol : float
    #    夏普率，为组合收益率除以波动率，此处不涉及无风险收益率资产
    #
    weights = np.array(weights)
    pret = np.sum(rets.mean() * weights) * 252
    pvol = np.sqrt(np.dot(weights.T, np.dot(rets.cov() * 252, weights)))
    return np.array([pret, pvol, pret / pvol])
#  以上是描述某投资组合是这样的一个函数，即输入权重分配，输出该组合的收益率、波动率和夏普率

print '准备函数完毕,输出最优组合1:'
#  下面利用上面定义函数求最大夏普率的投资组合（即求出权重）：
def min_func_sharpe(weights):
    return -statistics(weights)[2]
bnds = tuple((0, 1) for x in range(number_of_assets))                          #  边界条件bnds为： 每个权重需要在0到1之间
cons = ({'type': 'eq', 'fun': lambda x: np.sum(x) - 1})                        #  约束条件cons为： 权重之和为1
opts = sco.minimize(min_func_sharpe, number_of_assets * [1. / number_of_assets,],
                    method='SLSQP',  bounds=bnds, constraints=cons)         
# 优化选择结果opts['x']即为最大夏普率的投资组合的权重分配(保留三位有效数字）
print opts['x'].round(3)
print statistics(opts['x']).round(3)    # statistics(opts['x']).round(3)可获得最大夏普率的投资组合的收益率、波动率和夏普率
print '以0.54%权重买000651格力电器，94.6%权重买入600519贵州茅台，这样的组合年化收益率为70.2%，波动率为21.5%，夏普率为3.272'

#下面求最小方差组合：
print ''
print '输出最小方差组合2：'
def min_func_variance(weights):
    return statistics(weights)[1] ** 2
optv = sco.minimize(min_func_variance, number_of_assets * [1. / number_of_assets,],
                    method='SLSQP', bounds=bnds, constraints=cons)
print optv['x'].round(3)                # 最小方差结果optv['x']即为最小方差组合的权重分配(保留三位有效数字）
print statistics(optv['x']).round(3)    # statistics(optv['x']).round(3)可以获得最小方差的投资组合的收益率、波动率和夏普率


def min_func_port(weights):
    return statistics(weights)[1]
target_returns = np.linspace(0.3, 0.7, 50)   
target_volatilities = []                       
for tret in target_returns:
    cons = ({'type': 'eq', 'fun': lambda x:  statistics(x)[0] - tret},{'type': 'eq', 'fun': lambda x:  np.sum(x) - 1})
    res = sco.minimize(min_func_port, number_of_assets * [1. / number_of_assets,], method='SLSQP',bounds=bnds, constraints=cons)
    target_volatilities.append(res['fun'])

# 下面就可以通过散点图找出有效边界：
plt.figure(figsize=(15, 8))
# 圆点为随机资产组合
plt.scatter(portfolio_volatilities, portfolio_returns,
            c=portfolio_returns / portfolio_volatilities, marker='o')
# 叉叉为有效边界
plt.scatter(target_volatilities, target_returns,
            c=target_returns / target_volatilities, marker='x')
# 红星为夏普率最大值的资产组合
plt.plot(statistics(opts['x'])[1], statistics(opts['x'])[0],
         'r*', markersize=15.0)

plt.grid(True)
plt.xlabel('expected volatility')
plt.ylabel('expected return')
plt.colorbar(label='Sharpe ratio')
plt.show()





```
