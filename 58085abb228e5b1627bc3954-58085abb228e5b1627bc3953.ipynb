## 2.情景模型的因子加权矩阵
- ### 选择了情景分层因子之后，由于同样的因子在不同的分层中的预测能力和预测稳定程度存在着显著的差别。
- ### 因此，在不同的情景分层中，我们需要给同样的因子不同的权重。
- ### 赋予因子权重的依据是过去N个月的IR，即因子Rank_IC的均值除以标准差。
- ### 当然，这里也可以采用[基于因子IC的多因子模型](https://uqer.datayes.com/community/share/57b540ef228e5b79a4759398)里因子合成的办法，不过显得很复杂，这里就采用简单的因子IR作为标准
- ### 下面就是动手实践了，仍然要用到上次的一些函数，用于计算因子在一段时间内分层后的因子rank_IC,细节就不再赘述了

```python
class factor_perf_on_basical_factor(object):
    
    def __init__(self,basical_factor,factor,begindate,enddate,frequency='monthly',universe_name='A',dynamic=False):
        ###----传入参数----
        self.begindate = begindate  #开始时间
        self.enddate = enddate      #结束时间
        self.frequency = frequency  #数据频率，只要日度，周度，月度三个选择
        self.universe_name = universe_name  #股票收益溢价的股票池
        self.basical_factor = basical_factor #基本因子，用于分层
        self.factor = factor   #想要查看因子表现的因子名
        self.dynamic = dynamic   #universe是否使用动态
        ###----自定义参数----

        self.datelist = self.get_datelist()  #传入开始和结束日期，得到交易日列表list
        
    
    ###得到日期列表
    def get_datelist(self):   
        from CAL.PyCAL import *  #从dataapi调取交易日序列
        data=DataAPI.TradeCalGet(exchangeCD=u"XSHG",beginDate=self.begindate,endDate=self.enddate,field=['calendarDate','isWeekEnd','isMonthEnd'])
        if self.frequency == 'daily':   #按照规定的频率处理得到的交易日序列
            #self.M = 250
            date= map(lambda x: x[0:4]+x[5:7]+x[8:10], data['calendarDate'].values.tolist())
        if self.frequency == 'weekly':
            #self.M = 52
            data = data[data['isWeekEnd'] == 1]
            date= map(lambda x: x[0:4]+x[5:7]+x[8:10], data['calendarDate'].values.tolist())
        if self.frequency == 'monthly':
            #self.M = 12
            data = data[data['isMonthEnd'] == 1]
            date= map(lambda x: x[0:4]+x[5:7]+x[8:10], data['calendarDate'].values.tolist())
        return date
    
    #按basical_factor把股票分为数量相同的两组（大小），传入分组时的时间点
    def get_group(self,breakpoint):
        factor_filed = 'secID,'+self.basical_factor
        if self.dynamic:          #调用dataAPI得到股票代码及股票分层因子的值，如果是动态universe,则使用动态
            data = DataAPI.MktStockFactorsOneDayGet(tradeDate=breakpoint,secID=set_universe(self.universe_name,breakpoint),field=factor_filed).dropna()
        else:    
            data = DataAPI.MktStockFactorsOneDayGet(tradeDate=breakpoint,secID=set_universe(self.universe_name),field=factor_filed).dropna()
        data.sort(data.columns[1],ascending=True,inplace=True)     #按分层因子值从小到大排序
        secIDs = data['secID'].tolist()
        group_low = secIDs[:np.round(len(data))/2]        #分组,因子值小的哪一组股票为low,高的为high
        group_high = secIDs[np.round(len(data))/2:]
        return group_low,group_high
    
    #定义函数计算股票收益序列，传入时间，股票代码，得到收益序列
    def get_return(self,secIDs,date):
        index = self.datelist.index(date)
        data1 = DataAPI.MktEqudAdjGet(tradeDate=self.datelist[index-1],secID=secIDs,field=u"secID,closePrice").dropna() #前一期股票收盘价
        data1.set_index('secID',inplace=True)
        data2 = DataAPI.MktEqudAdjGet(tradeDate=date,secID=secIDs,field=u"secID,closePrice").dropna()  #当期股票收盘价
        data2.set_index('secID',inplace=True)
        data = pd.concat([data1,data2],axis=1).dropna()
        data.columns=['return','closeprice']
        data['return'] = data['closeprice']*1.0/data['return']-1    #得到收益序列
        del data['closeprice']
        returns = standardize(neutralize(winsorize(data['return'].dropna().to_dict()),date))    #对个股收益作标准化，去极值，中性化处理
        data['return'][returns.keys()] = returns.values()
        return data
    
    #得到因子值
    def get_factor(self,secIDs,date):
        date = self.datelist[self.datelist.index(date)-1]
        factor_filed = 'secID,'+self.factor
        data = DataAPI.MktStockFactorsOneDayGet(tradeDate=date,secID=secIDs,field=factor_filed).dropna()  #得到上期因子值
        data.set_index('secID',inplace=True)      #设置index为股票代码
        for i in data.columns:
            tmp = standardize(neutralize(winsorize(data[i].to_dict()),date))  #对因子值作标准化，去极值，中性化处理
            data[i][tmp.keys()] = tmp.values()
        return data
    
    #计算因子的rank_IC
    def get_rankIC(self):
        low={}
        high={}
        tmp = DataAPI.MktStockFactorsOneDayGet(tradeDate='20160930',ticker='000001',field=self.factor)
        for i in tmp.columns:
            low[i] = []                    #用于存储low那一组股票的rankIC
            high[i] = []                    #用于存储high那一组股票的rankIC
        for j in range(1,len(self.datelist)):   #对时间做循环，得到每个时间点的rankIC
            date = self.datelist[j]
            group_low,group_high = self.get_group(date)         #分组
            returns_low = self.get_return(group_low,date)
            returns_high = self.get_return(group_high,date)     #得到当期收益序列
            factor_low = self.get_factor(group_low,date)
            factor_high = self.get_factor(group_high,date)      #得到上期因子序列
            returns_low = pd.concat([returns_low,factor_low],axis=1).dropna()    #对应起来
            returns_high = pd.concat([returns_high,factor_high],axis=1).dropna()
            for t in tmp.columns:
                tmplow,_=st.spearmanr(returns_low['return'],returns_low[t])
                tmphigh,_=st.spearmanr(returns_high['return'],returns_high[t])
                low[t].append(tmplow)
                high[t].append(tmphigh)
        return low,high,tmp.columns
```

- ### 下面的函数就是根据计算出的因子Rank_IC来计算因子加权矩阵
- ### 仔细说明一下每个参数的意义

> N:前N期的Rank_IC来计算因子IR

> basical_factors：情景分层因子，对每一个因子都要作为分层的标准计算IR

> factors：Alpha因子，探查其在分层因子分层后的不同基本面上的权重

> begindate：开始时间

> enddate：结束时间

> frequency：数据频率，默认为月度

> universe_name：股票池名称

> dynamic是否动态universe

- ### 因子权重与因子IR成正比，我在这里就简单的把IR的值作为权重了

```python
def IR_weight(N,basical_factors,factors,begindate,enddate,frequency='monthly',universe_name='A',dynamic=False):
    result={}   #用于存储变量
    #wrong={}
    for basical_factor in basical_factors:   #每一个因子进行分层
        perf = factor_perf_on_basical_factor(basical_factor,factors,begindate,enddate,frequency,universe_name,dynamic)
        datelist = perf.datelist[1:]         #每个日期
        low,high,factor_names = perf.get_rankIC()    #得到分层后因子rankIC的值
        result[basical_factor] = {}          #用于存储结果
        if len(datelist)<N:                #长度不够时，报错
            print 'lenth of data is not enough,please reset date'
            return
        for i in range(N-1,len(datelist)):
            IR_low=[]
            IR_high=[]
            weight=pd.DataFrame(index=factor_names,columns=[basical_factor+'_low',basical_factor+'_high'],data=0)  #用于存储结果
            for factor in factor_names:  #计算因子IR
                mean_low=np.array(low[factor][i-(N-1):i]).mean()
                std_low=np.array(low[factor][i-(N-1):i]).std()
                mean_high=np.array(high[factor][i-(N-1):i]).mean()
                std_high=np.array(high[factor][i-(N-1):i]).std()
                IR_low.append(mean_low/std_low)
                IR_high.append(mean_high/std_high)
            weight[weight.columns[0]] = IR_low
            weight[weight.columns[1]] = IR_high
            '''
            if np.array(IR_low).sum() == 0:    #计算因子权重
                weight[weight.columns[0]] = IR_low
            else:
                weight[weight.columns[0]] = map(lambda x:(x*1.0/np.array(IR_low).sum())*100,IR_low)
            if np.array(IR_high).sum() == 0:
                weight[weight.columns[1]] = IR_high
            else:
                weight[weight.columns[1]] = map(lambda x:(x*1.0/np.array(IR_high).sum())*100,IR_high)
            '''
            result[basical_factor][datelist[i]] = weight.T
    result_daily=pd.DataFrame()    #重新整理结果
    for j in datelist[N-1:]:      #对时间循环
        tmp=pd.DataFrame()
        for t in basical_factors:    #对分层因子循环
            tmp = pd.concat([tmp,result[t][j]],axis=0)
        indexs = [[j]*len(tmp),tmp.index.tolist()]
        tmp.index = pd.MultiIndex.from_tuples(zip(*indexs))
        result_daily = pd.concat([result_daily,tmp],axis=0)
    return result_daily
```

# 2.情景分层因子的表现
- ### 正如前文所说，每个因子在不同的分层下有不同的权重，而这里我们有5个情景分层因子，遗憾的是这不像行业划分那样两两不交
- ### 那我们不妨对每个情景分层因子都探查一下
- ### 策略逻辑：对于一个分层因子，每次调仓时都把股票按照分层因子的值先分为low和high两层，对于每一层按照之前算出的因子加权矩阵里的权重对九个alpha因子的值求和，作为股票的打分，再使用simple_long_only来构建股票的权重，其余设置和之前相同
- ### 这里使用快速回测框架进行回测，帮助里有详细介绍

```python
#快速回测
# 可编辑部分与 strategy 模式一样，其余部分按本例代码编写即可

# -----------回测参数部分开始，可编辑------------
# 导入包
from CAL.PyCAL import *


start = '2014-09-30'                       # 回测起始时间
end = '2016-10-15'     # 回测结束时间
universe = DynamicUniverse('000906.ZICN')     # 股票池
benchmark = 'HS300'                       # 策略参考标准
capital_base = 10000000                     # 起始资金

# 日期处理相关
cal = Calendar('China.SSE')
period = Period('-1B')
factor_name = 'PB,ETOP,PS,GrossIncomeRatio,PE,NetProfitGrowRate,VOL20,EMA5,HBETA'
basical_factors=['LCAP','PB','NetAssetGrowRate','ROE','VOL60']


# ---------------回测参数部分结束----------------

# 把回测参数封装到 SimulationParameters 中，供 quick_backtest 使用
sim_params = quartz.SimulationParameters(start, end, benchmark, universe, capital_base, max_history_window=max_history_window)
# 获取回测行情数据
data = quartz.get_backtest_data(sim_params)
# 运行结果
results = {}
# 调整参数快速回测
for x in range(5):

    # -----------策略逻辑部分开始，可编辑------------

    # refresh_rate 的值和回测行情数据无关，所以也可以在参数调优中修改
    # 类似的还有 commission 和 slippage
    refresh_rate = 1
    # commission = ...
    # slippage = ...

    def initialize(account):
        pass
    def handle_data(account):
        today = account.current_date
        today = Date.fromDateTime(account.current_date)  # 向前移动一个工作日
        yesterday = cal.advanceDate(today, period)
        yesterday = yesterday.toDateTime().strftime('%Y%m%d')
        if yesterday in datelist:
            
            groupdata = DataAPI.MktStockFactorsOneDayGet(tradeDate=yesterday,secID=account.universe,field='secID,'+basical_factors[x]).dropna()
            groupdata.set_index('secID',inplace=True)
            groupdata.sort([basical_factors[x]],ascending=True,inplace=True)
            low = groupdata.index[:int(len(groupdata)/2)]
            high = groupdata.index[int(len(groupdata)/2):]
        
            low_factor = DataAPI.MktStockFactorsOneDayGet(tradeDate=yesterday,secID=low,field='secID,'+factor_name)
            low_factor.set_index('secID',inplace=True)
            high_factor = DataAPI.MktStockFactorsOneDayGet(tradeDate=yesterday,secID=high,field='secID,'+factor_name)
            high_factor.set_index('secID',inplace=True)
        
            # 构建组合score矩阵
            low_Score = pd.DataFrame(index=low_factor.index, columns=low_factor.columns, data=0)
            for name in low_factor.columns:
                signal1 = standardize(neutralize(winsorize(low_factor[name].to_dict()),yesterday))   # 因子处理
                low_Score[name][signal1.keys()] = signal1.values()
            weightlow = np.array(my_trial.ix[yesterday].ix[basical_factors[x]+'_low'])
            low_Score['total_score'] = np.dot(low_Score, weightlow)
            
            high_Score = pd.DataFrame(index=high_factor.index, columns=high_factor.columns, data=0)
            for name in high_factor.columns:
                signal2 = standardize(neutralize(winsorize(high_factor[name].to_dict()),yesterday))   # 因子处理
                high_Score[name][signal2.keys()] = signal2.values()
            weighthigh = np.array(my_trial.ix[yesterday].ix[basical_factors[x]+'_high'])
            high_Score['total_score'] = np.dot(high_Score, weighthigh)
        
            total_score = pd.concat([low_Score,high_Score],axis=0)
            wts = simple_long_only(total_score['total_score'].to_dict(),yesterday)  # 组合构建函数
        
            total_score['wts'] = np.nan
            total_score['wts'][wts.keys()] = wts.values()
            total_score = total_score[~np.isnan(total_score['wts'])]
            #factor.set_index('secID', inplace=True)
            #factor.drop(factor_name, axis=1, inplace=True)
        
            # 先卖出
            sell_list = account.security_position
            for stk in sell_list:
                order_to(stk, 0)

            # 再买入
            buy_list = total_score.index
            total_money = account.reference_portfolio_value
            prices = account.reference_price 
            for stk in buy_list:
                if np.isnan(prices[stk]) or prices[stk] == 0:  # 停牌或是还没有上市等原因不能交易
                    continue
                order(stk, int(total_money * total_score.loc[stk]['wts'] / prices[stk] /100)*100)
        else:
            return

    # ---------------策略逻辑部分结束----------------

    # 把回测逻辑封装到 TradingStrategy 中，供 quick_backtest 使用
    strategy = quartz.TradingStrategy(initialize, handle_data)
    # 回测部分
    bt, perf = quartz.quick_backtest(sim_params, strategy, data=data, refresh_rate=refresh_rate)
    # commission = commission,
    # slippage = slippage

    # 保存运行结果
    results[basical_factors[x]] = perf
```

```python
#汇总结果
result = pd.DataFrame()
result['情景分层因子'] = basical_factors
result['年化收益'] = [results[i]['annualized_return'] for i in basical_factors]
result['最大回撤'] = [results[i]['max_drawdown'] for i in basical_factors]
result['收益波动率'] = [results[i]['volatility'] for i in basical_factors]
result['信息比率'] = [results[i]['information_ratio'] for i in basical_factors]
result['夏普比率'] = [results[i]['sharpe'] for i in basical_factors]
```

## 3.动态情景多因子策略
- ### 下面就是考虑如何将不同分层的因子权重统一起来使用了
- ### 一种简单的想法就是，把不同情景分层因子所得到的alpha因子权重直接简单求和
- ### 策略逻辑：每次调仓时,alpha因子的权重按照之前计算出的因子加权矩阵，把不同分层下的权重直接求和（即，每个日期的权重矩阵对列求和作为因子的权重），然后因子值加权求和作为股票的打分，再使用simple_long_only来构建股票的权重，其余设置和之前相同

- ### 从结果上来看，效果比之前单个情景分层因子来构建策略的效果都更好了，收益风险比更高了
- ### 为了说明策略是否真的有效，下面使用快速回测来看看分位组合的效果

```python
#快速回测
# 可编辑部分与 strategy 模式一样，其余部分按本例代码编写即可

# -----------回测参数部分开始，可编辑------------
# 导入包
from CAL.PyCAL import *


start = '2014-09-30'                       # 回测起始时间
end = '2016-10-15'     # 回测结束时间
universe = DynamicUniverse('000906.ZICN')     # 股票池
benchmark = 'HS300'                       # 策略参考标准
capital_base = 10000000                     # 起始资金

# 日期处理相关
cal = Calendar('China.SSE')
period = Period('-1B')
basical_factors='LCAP，PB,NetAssetGrowRate,ROE，VOL60'


# ---------------回测参数部分结束----------------

# 把回测参数封装到 SimulationParameters 中，供 quick_backtest 使用
sim_params = quartz.SimulationParameters(start, end, benchmark, universe, capital_base, max_history_window=max_history_window)
# 获取回测行情数据
data = quartz.get_backtest_data(sim_params)
# 运行结果
results_quantile = {}
# 调整参数快速回测
for x in range(5):

    # -----------策略逻辑部分开始，可编辑------------

    # refresh_rate 的值和回测行情数据无关，所以也可以在参数调优中修改
    # 类似的还有 commission 和 slippage
    refresh_rate = 1
    # commission = ...
    # slippage = ...

    def initialize(account):
        pass
    def handle_data(account):
        today = account.current_date
        today = Date.fromDateTime(account.current_date)  # 向前移动一个工作日
        yesterday = cal.advanceDate(today, period)
        yesterday = yesterday.toDateTime().strftime('%Y%m%d')
        if yesterday in datelist:
            
            
            factor = DataAPI.MktStockFactorsOneDayGet(tradeDate=yesterday,secID=account.universe,field='secID,'+factor_name).dropna()
            factor.set_index('secID',inplace=True)
        
            # 构建组合score矩阵
            Total_Score = pd.DataFrame(index=factor.index, columns=factor.columns, data=0)
            for name in factor.columns:
                signal = standardize(neutralize(winsorize(factor[name].to_dict()),yesterday))   # 因子处理
                Total_Score[name][signal.keys()] = signal.values()
            weight = np.array(my_trial.ix[yesterday].sum())
            factor['total_score'] = np.dot(Total_Score, weight)
            quantile1ow=np.percentile(factor['total_score'],x*20)
            quantilehigh=np.percentile(factor['total_score'],(x+1)*20)
            factor=factor[(factor['total_score']>=quantile1ow) & (factor['total_score']<=quantilehigh)]
        
            # 先卖出
            sell_list = account.security_position
            for stk in sell_list:
                order_to(stk, 0)

            # 再买入
            buy_list = factor.index
            total_money = account.reference_portfolio_value
            prices = account.reference_price 
            for stk in buy_list:
                if np.isnan(prices[stk]) or prices[stk] == 0:  # 停牌或是还没有上市等原因不能交易
                    continue
                order(stk, int(total_money*1.0 /len(buy_list)/ prices[stk] /100)*100)
        else:
            return

    # ---------------策略逻辑部分结束----------------

    # 把回测逻辑封装到 TradingStrategy 中，供 quick_backtest 使用
    strategy = quartz.TradingStrategy(initialize, handle_data)
    # 回测部分
    bt, perf = quartz.quick_backtest(sim_params, strategy, data=data, refresh_rate=refresh_rate)
    # commission = commission,
    # slippage = slippage

    # 保存运行结果
    results_quantile[x] = perf
```
