    **一、估值维度**
    （1）企业价值倍数：
    企业价值倍数衡量的是针对整个公司的资本能够创造多少实际收益。企业价值倍数与市盈率类似，也是结合公司的盈利能力与公司价值考量公司股价相对公司本身内在价值的比率。但是，企业价值倍数在多个维度上优于市盈率。首先，EBITDA把一些非公司个体成本加回来，使得这些因行业不同而不同的支出不会影响公司的估值水平。其次，由于EBITDA包含了很多公司常用来粉饰财务报表的项目，因此被操纵的可能性很小，有利于投资者排除很多干扰，能够更准确地判断公司核心业务的经营状况。最后，由于EBIITDA包含了几乎所有公司运营所需变动的账目，因此每年EBITDA的变动幅度并不会像净利润那样出现很大的波动。
    特点：适合在不同行业进行比较；不适用于某些行业内部的比较（如航空运输业）；没有衡量公司的成长速度和营收规模，因此不能单一使用该指标作最优评价。
    计算公式：企业价值倍数 = 企业价值/EBITDA 
    其中，企业价值 = 市值 + 负债（-现金）；EBITDA=净利润+所得税+固定资产折旧+无形资产摊销+长期待摊费用摊销+偿付利息所支出的现金。
    实证：Gray and Vogel(2012)利用1991-2010年的数据，验证了企业价值倍数能否作为投资决策这一命题。当他们剔除了市值最小的10%公司金融股、公用事业股并根据企业价值倍数的大小将股票分为5组之后，发现企业价值倍数最低一组的年化收益为17.66%，而企业价值倍数最高一组的年化收益率仅为7.97%，两组平均收益之差高达9.69%，这也使得企业价值成为区分度最高的指标。
    
    （2）股息率：
    股息率衡量的是投资者每年能够从公司实际运营中拿到的分红占自身投资资金的比率。从今年年初至今来看，小市值因子失效，相反股息率因子表现得非常出色，上证50屡创新高，证明今年市场风格偏好绩优蓝筹股。
    计算公式：股息率 = 每股分红/每股价格
    实证：Tortoriello(2009)使用20年(1988-2007年)的美国股票数据，发现尽管股息率最高的一组可以带来2.4%的超额收益，但是该超额收益比起其他指标略显不足；而且每个分组之间的差异非常小。显然，该指标并不是一个非常好的投资标准。（的确，在加了股息率因子后，估值维度因子的五分位表现均有了一定程度的下降）
    
    （3）市盈率：
    从投入与回报的角度来看，市盈率表示的是用税后收益需要多少年的累积才能达到当前的股价。通过这样的方式，市盈率提供了一个衡量股价高低的尺度。
    计算公式：市盈率 = 每股市场价格/每股税后收益
    实证：Anderson and Brooks(2006)使用1975-2003年的数据，发现当剔除负市盈率股票后，市盈率最低一组与市盈率最高一组之间存在年化8.31%的收益率差异；而当使用长达8年平均市盈率作为分组标准时，市盈率最高一组与市盈率最低一组的年化收益差异达到11.62%。
    
    （4）市净率：
    市净率是一个广泛应用于证券分析的价值衡量指标。从估值角度来看，市场价值不仅包括账面价值，还包括利用账面价值资产所创造的价值。
    计算公式：市净率 = 每股价格/每股净资产
	实证：Tortoriello(2009)把市净率与公司的盈利能力ROIC结合起来，显著地提高了股票组合的收益率。他使用了1988-2007年的每股数据，发现市净率最低的20%股票组合可以带来14.5%的年化收益率，同期标准普尔500指数可以带来12.9%的年化收益率，超额收益为1.6%。而当他使用1990-2007年的数据时，将市净率与盈利能力ROIC结合，发现市净率最低、ROIC最高的一组（最被低估的一组）可以带来18.7%的年化收益率，超额收益为6.7%。

	（5）市销率：
	市销率是公司的市值与其销售额（营业收入）的比值。它表示的是一种类似于衡量流量的方法，即公司的市值是营业收入的多少倍，公司需要多少年才能够赚得市值那么多的收入。
	计算公式：市销率 = 市场价值/营业收入
	实证：Senchack and Martin(1987)使用1975-1984年的季度回报数据，发现市销率越低，其组合平均收益率越高。市销率最低一组的平均收益率最高，达到0.28，但是其风险也最高，高达1.178；市销率最高一组的年收益只有约0.18，但是风险并不低。

```python
from quartz_extensions import *
start = '20100819'                       # 回测起始时间
end = '20170819'                         # 回测结束时间
benchmark = 'ZZ500'                      # 策略参考标准
universe = set_universe('ZZ500')         # 证券池，支持股票和基金
refresh_rate = 'month'                     # 调仓频率，表示每周五结束之后生成调仓
delay_number = 5                         # 信号最大衰减周期
quantile_number = 5                      # 分位分组的组数
function_list = ["winsorize", "neutralize", "standardize"]    # 信号处理函数列表，从前到后进行处理
industry_type = 'SW1'                    # 行业分类类型

## 自定义因子函数
def foo(data, dependencies=['MktValue','NetDebt','FCFF','EBITDA','ROIC','PE','PB','PS','DividendPaidRatio'], max_window=1):
    EnterpriseValueMultiple = 1/((data['MktValue'].mean()+data['NetDebt'].mean()-data['FCFF'].mean())/data['EBITDA'].mean())
    EnterpriseValueMultiple = EnterpriseValueMultiple[EnterpriseValueMultiple>0]
    ROIC = data['ROIC'].mean()
    PE = 1/data['PE'].mean()
    PE = PE[PE>0]
    PB = 1/data['PB'].mean()
    PB = PB[PB>0]
    PS = 1/data['PS'].mean()
    PS = PS[PS>0]
    DividendPaidRatio = data['DividendPaidRatio'].mean()
    Valuation = EnterpriseValueMultiple + ROIC + PE + PB + PS + DividendPaidRatio
    # print Valuation
    return Valuation

## 注册因子名称
def initialize(account):
    a = Signal('Valuation', foo)
    account.signal_generator = SignalGenerator(a)

## 调用自定义因子分析函数
create_factor_tear_sheet('Valuation', start=start, end=end, benchmark=benchmark, universe=universe, refresh_rate=refresh_rate, delay_number=delay_number, quantile_number=quantile_number, function_list=function_list, industry_type=industry_type, initialize=initialize)
```

    **二、质量维度**
	（1）盈利能力：
	投资资本回报率ROIC，该指标能够识别出公司真正具备生产力的资产回报。如果投资资本回报率很高，理论上公司就更有可能具备成长性并更有可能在未来获得超额收益。
	计算公式：ROIC = 息税前利润x（1-所得税税率）/（固定资产+无形资产+流动资产-流动负债-现金）
	实证：刘俏(2014)发现，美国上市公司在1961-2001年时间的投资资本回报率年平均值高达11.6%，这与投资者同期能够获得的平均回报相当。而中国上市公司在1998-2012年，投资资本回报率的算术平均数只有3%，按总资产加权平均值为4.1%，远低于美国上市公司。

	1.盈利能力的持续性通常可以采用公司的长期平均资产回报率：ROA5
	2.盈利能力的成长性可以采用公司的毛利增长率：DEGM 
	3.盈利能力的稳定性可以采用：平均毛利率/毛利率标准差

	（2）经营效率：
	经营资产周转率ATO是最基础和经典的经营效率度量方法，受到广泛认可，使用也最为普遍。资产周转率取决于资产使用和管理效率，包括固定资产使用效率、存货管理及其他营运资本管理。
	计算公式：ATO = 营业收入/平均净经营资产
	实证：Soliman教授认为，如果当前RNOA的提高主要源自于ATO，那么未来ATO及RNOA会比预期更高。

	（3）盈余质量：
	应计利润是衡量盈余操纵最经典的指标，是公司净利润(Earnings)与经营性净现金流量(CFO)的差异。
	计算公式：Accrual = Earnings - CFO
	实证：Sloan(1996)发现，在每一年年初，根据总应计利润对公司进行排序，买入应计利润最低的10%的股票并卖空应计利润最高的10%的股票，分别选择在1年、2年和3年后平仓；他最终发现，1年后平仓的年化对冲收益为10.4%，2年为4.8%，3年为3.8%。这种基于应计利润的错误定价被称为“应计异象”。

	识别财务操纵的M-score模型，Beneish(1999)筛选出的八个指标：应收账款指数，毛利率指数，资产质量指数，营业收入指数，折旧率指数，销售管理费用指数，财务杠杆指数，应计系数。每个指标的数值越大，在Beneish看来都意味着更大的盈余操纵程度。

	动态财务困境预测模型Campbell财务困境模型，Campbell(2008)选择了八个指标：盈余资产市价比，总负债资产市价比，现金资产市价比，现金资产市价比，月度超额收益，相对规模，市值账面价值比，股票价格，过去3个月股票年化波动率。

	以上这两个模型，我认为可以用来排除股票池中的“地雷”，避免买入如乐视网等股票。

	（4）投融资决策：
	1.超额资本支出与超额收益之间存在负相关关系--代理问题存在
	2.股权融资后，长期观察到公司的负超额收益--融资优序理论
	3.股份被公司回购，往往有正的超额收益
	4.公司在发起股利之后会有正的超额收益，缺失股利之后会有负的超额收益
	5.总资产增长率与公司的超额收益之间存在显著的负相关关系

	（5）无形资产：
	在商业变革的环境中，公司越来越多地进行无形资产投资，然而会计准则对于无形资产当期费用化的规定，使得财务报表信息的价值相关性逐渐消失。对无形资产价值的认知和理解，是恢复财务信息价值相关性、重新发现公司价值的重要环节。
	1.由于投资者没有充分意识到研发支出的价值，研发支出程度越大的公司在未来能够获得越高的超额收益。
	2.从动态变化的角度来看，当期研发支出增幅越大，投资者对研发支出价值的低估程度也就越严重，因此研发支出增幅越大的公司在未来也能获得越高的超额收益。
	3.在研发支出程度一定的情形下，公司创新效率越高，单位研发支出转化的专利技术数量越多，未来增加的企业价值就越多。

	在此也希望优矿未来能加上研发费用这个因子

```python
from quartz_extensions import *
start = '20100819'                       # 回测起始时间
end = '20170819'                         # 回测结束时间
benchmark = 'ZZ500'                      # 策略参考标准
universe = set_universe('ZZ500')         # 证券池，支持股票和基金
refresh_rate = 'month'                     # 调仓频率，表示每周五结束之后生成调仓
delay_number = 5                         # 信号最大衰减周期
quantile_number = 5                      # 分位分组的组数
function_list = ["winsorize", "neutralize", "standardize"]    # 信号处理函数列表，从前到后进行处理
industry_type = 'SW1'                    # 行业分类类型

## 自定义因子函数
def foo(data, dependencies=['RevenueTTM','TotalFixedAssets','ROIC','ROA','DEGM','GrossIncomeRatio','NetProfitTTM','NetOperateCFTTM','ARTRate','RetainedEarnings','TotalAssets','TotalAssetGrowRate','EPS','NetInvestCFTTM','OperatingRevenueGrowRate','IntangibleAssetRatio'], max_window=20):
    ATO = data['RevenueTTM'].mean()/data['TotalFixedAssets'].mean()
    ROIC = data['ROIC'].mean()
    ROA = data['ROA'].mean()
    DEGM = data['DEGM'].mean()
    GrossIncomeRatio = data['GrossIncomeRatio'].mean()/data['GrossIncomeRatio'].std()
    Accrual = 1/(data['NetProfitTTM'].mean()-data['NetOperateCFTTM'].mean())
    ARTRate = data['ARTRate'].mean()
    RET = data['RetainedEarnings'].mean()/data['TotalAssets'].mean()
    TotalAssetGrowRate = 1/data['TotalAssetGrowRate'].mean()
    EPS = 1/data['EPS'].mean()
    NetInvestCFTTM = data['NetInvestCFTTM'].mean()
    OperatingRevenueGrowRate = data['OperatingRevenueGrowRate'].mean()
    IntangibleAssetRatio = data['IntangibleAssetRatio'].mean()
    
    Quality = ATO + ROIC + ROA + DEGM + GrossIncomeRatio + Accrual + ARTRate + RET + TotalAssetGrowRate + EPS + NetInvestCFTTM + OperatingRevenueGrowRate + IntangibleAssetRatio
    
    return Quality

## 注册因子名称
def initialize(account):
    a = Signal('Quality', foo)
    account.signal_generator = SignalGenerator(a)

## 调用自定义因子分析函数
create_factor_tear_sheet('Quality', start=start, end=end, benchmark=benchmark, universe=universe, refresh_rate=refresh_rate, delay_number=delay_number, quantile_number=quantile_number, function_list=function_list, industry_type=industry_type, initialize=initialize)
```

	**三、市场参与者信号**
	证券分析师：
	股票评级等级是分析师根据所持信息和自己对预期的判断，经分析后形成的、对于股票投资建议给出的态度和结果。
	一只股票的评级反映了卖方分析师对公司内在价值相比市场价值的评价与判断。相对于股票的市场价值而言，内在价值越高，股票评级就越乐观，因为股票被低估了，其未来市场价值会向内在价值靠拢。股票评级减少了投资者获取信息的成本，是投资者决策的重要依据，投资者可以买入评级高的股票以获取超额收益。
	实证：Li(2005)的研究表明，如果在1994-2000年买入评级为“强烈买入”的股票并赋予相同的权重，可以获得5%的年化超额收益；卖出评级为“卖出”或“逊于大盘”的股票，可以获得3.6%的年化收益。
	其他一些研究则表明，选择基本面好的公司可以改善投资效益。Jegadeesh et al.(20004)指出，股票评级因子结合好的公司基本面特征可以提高收益，如较高的动量特征、较低的市盈率和市净率、较低的应计项目、较低的增长率和较低的资本支出。如果投资者在1985-1998年选取基本面好的公司并赋予相同权重，买入（卖出）一致评级高（低）的公司可以获得5.6%的市场调整收益，而买入（卖出）评级调高（调低）的公司可以获得7.2%的市场调整收益。
	从我的角度来看，财务因子是反映了公司过去的历史信息，而分析师因子是反映了公司当前以及未来的信息（因为分析师一般要去进行实地调研，以前我也曾是一名行业研究员，去公司调研与董秘聊天的时候，总会有一些财务报表上没有的信息），因此分析师因子具有“增量”贡献。

```python
from quartz_extensions import *
start = '20100819'                       # 回测起始时间
end = '20170819'                         # 回测结束时间
benchmark = 'ZZ500'                      # 策略参考标准
universe = set_universe('ZZ500')         # 证券池，支持股票和基金
refresh_rate = 'month'                     # 调仓频率，表示每周五结束之后生成调仓
delay_number = 5                         # 信号最大衰减周期
quantile_number = 5                      # 分位分组的组数
function_list = ["winsorize", "neutralize", "standardize"]    # 信号处理函数列表，从前到后进行处理
industry_type = 'SW1'                    # 行业分类类型

## 自定义因子函数
def foo(data, dependencies=['DAREC','DAREV','DASREV','EPIBS','EgibsLong','FEARNG','FY12P','GREC','GREV','GSREV','REC','SFY12P','REVS20','REVS60'], max_window=1):
    DAREC = data['DAREC'].mean()
    DAREC = DAREC[DAREC>0]
    DAREV = data['DAREV'].mean()
    DASREV = data['DASREV'].mean()
    EPIBS = data['EPIBS'].mean()
    EgibsLong = data['EgibsLong'].mean()
    FEARNG = data['FEARNG'].mean()
    FY12P = data['FY12P'].mean()
    GREC = data['GREC'].mean()
    EPIBS = data['EPIBS'].mean()
    GREV = data['GREV'].mean()
    GSREV = data['GSREV'].mean()
    REC = data['REC'].mean()
    SFY12P = data['SFY12P'].mean()
    
    REVS20 = data['REVS20'].mean()
    REVS60 = data['REVS60'].mean()

    AnalystsExpectR = DAREC+DAREV+DASREV+EPIBS+EgibsLong+FEARNG+FY12P+GREC+EPIBS+GREV+GSREV+REC+SFY12P+REVS20+REVS60
    # print AnalystsExpectR
    return AnalystsExpectR

## 注册因子名称
def initialize(account):
    a = Signal('AnalystsExpectR', foo)
    account.signal_generator = SignalGenerator(a)

## 调用自定义因子分析函数
create_factor_tear_sheet('AnalystsExpectR', start=start, end=end, benchmark=benchmark, universe=universe, refresh_rate=refresh_rate, delay_number=delay_number, quantile_number=quantile_number, function_list=function_list, industry_type=industry_type, initialize=initialize)
```
