# 对于比赛提一个小小的建议

    根据观察我发现，比赛的排名中模拟盘的记分是按照  模拟交易－－－>自己策略的“累计受益率”计算的。    
    我认为问题出在这个累计受益率不是从比赛设定的模拟盘开始的时间：6月23日计算的。而是从提交策略的那一天计算的。
    这样的话存在一个问题，因为大家都知道6月23日之前的那一周大盘跌掉了接近 12%，  很多之前提交策略的人如果在6月22日的时候终止策略，再重新提交策略，那么就规避了这一周的巨大跌幅。对于没有这样操作（在正式模拟盘开始前(6月23日之前的那个周六，周日)终止策略重新提交）的选手而言，那一周12%的跌幅对策略的收益率造成了很大的影响。
    最为一个学计算机的同学，我知道可能通过按照选手提交日开始计算收益率(模拟盘)得分的方法，程序实现起来比较简单，但是我弱弱的建议一下，既然是比赛，可否按照这个规范：
    回测得分计算区间：   2012-12-01 ～ 2015-05-01   就是官方设定的回测区间。
    模拟盘得分计算区间： 2015-06-23 ～ 2015-07-24
    
    我知道，如果按照我的这个提议，工程师哥哥可能需要加班修改程序了，但是我认为这个计算方式比较合理。
