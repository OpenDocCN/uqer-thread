```python
flow=db['partyPct_y']-db['partyPct_x']
b=cgjl1.sort_values('partyPct',axis=0,ascending=False)
b.insert(4,'flow',flow)
ticker_list=b['ticker']
c=DataAPI.MktStockFactorsOneDayGet(tradeDate=u"20191126",secID=u"",ticker=ticker_list,field=u"REVS20",pandas="1")
d=DataAPI.MktStockFactorsOneDayGet(tradeDate=u"20191126",secID=u"",ticker=ticker_list,field=u"REVS10",pandas="1")
e=DataAPI.MktStockFactorsOneDayGet(tradeDate=u"20191126",secID=u"",ticker=ticker_list,field=u"REVS5",pandas="1")

b.insert(5,'REVS20',c)
b.insert(5,'REVS10',d)
b.insert(5,'REVS5',e)
b['flow']=b.flow.fillna(b.partyPct)
flow_radio=b['flow']/b['partyPct']
b.insert(5,'flow_radio',flow_radio)
b.sort_values('flow_radio',ascending=False,axis=0).head(100)
```
