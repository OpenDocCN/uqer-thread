```python
from datetime import datetime
import pandas as pd
df=DataAPI.MktIdxdGet(indexID=u"",ticker=u"000001",beginDate=u"20100101",endDate=u"20181231",exchangeCD=u"XSHE,XSHG",field=u"tradeDate,CHGPct",pandas="1")
a=pd.to_datetime(df['tradeDate']).tolist()
b=[]
for i in a:
    c=i.weekday()
    b.append(c)
df['xinqi']=b
df.set_index('tradeDate')
#周一上涨概率
df1=df[df['xinqi']==0]
print df1[df1['CHGPct']>0].count()/df1['CHGPct'].count()
#周二上涨概率
df1=df[df['xinqi']==1]
print df1[df1['CHGPct']>0].count()/df1['CHGPct'].count()
#周三上涨概率
df1=df[df['xinqi']==2]
print df1[df1['CHGPct']>0].count()/df1['CHGPct'].count()
#周四上涨概率
df1=df[df['xinqi']==3]
print df1[df1['CHGPct']>0].count()/df1['CHGPct'].count()
#周五上涨概率
df1=df[df['xinqi']==4]
print df1[df1['CHGPct']>0].count()/df1['CHGPct'].count()
```
