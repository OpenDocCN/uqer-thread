1. Theta格式
-----------------
***

**Theta格式**可以认为是显式格式与隐式格式的加权平均，在某些特殊参数取值下，Theta格式可以获得二阶的时间方向收敛速度（显式格式以及隐式格式在时间方向的收敛速度都只有一阶）。

2. 算法
------------------
***

```python
import numpy as np
import math
import seaborn as sns
from matplotlib import pylab
from CAL.PyCAL import *
font.set_size(15)
```

按照之前介绍的隐式差分格式的方法，用离散差分格式代替连续微分。但是我们这里增加一个新的参数 $\theta$ 用来做隐式格式与显示格式的加权平均：

\begin{align*}
&\frac{C_{j,k+1} - C_{j,k}}{\Delta \tau} = （1-\theta)[(r - \frac{1}{2}\sigma^2)\frac{C_{j+1,k+1} - C_{j-1,k+1}}{2\Delta x} + \frac{1}{2}\sigma^2 \frac{C_{j+1,k+1} - 2C_{j,k+1} + C_{j-1,k+1}}{\Delta x^2} - rC_{j,k+1}] \\\\[5pt]
&+ \theta [(r - \frac{1}{2}\sigma^2)\frac{C_{j+1,k} - C_{j-1,k}}{2\Delta x} + \frac{1}{2}\sigma^2 \frac{C_{j+1,k} - 2C_{j,k} + C_{j-1,k}}{\Delta x^2} - rC_{j,k}]\\\\[5pt]
&\Rightarrow \quad C_{j,k+1} - C_{j,k} - (1-\theta)[(r - \frac{1}{2}\sigma^2)\frac{\Delta \tau}{2\Delta x}(C_{j+1,k+1} - C_{j-1,k+1}) + \frac{1}{2}\sigma^2 \frac{\Delta \tau}{\Delta x^2}(C_{j+1,k+1} - 2C_{j,k+1} + C_{j-1,k+1}) - r\Delta \tau C_{j,k+1}] \\\\[5pt]
&= \theta[(r - \frac{1}{2}\sigma^2)\frac{\Delta \tau}{2\Delta x}(C_{j+1,k} - C_{j-1,k}) + \frac{1}{2}\sigma^2 \frac{\Delta \tau}{\Delta x^2}(C_{j+1,k} - 2C_{j,k} + C_{j-1,k}) - r\Delta \tau C_{j,k}], \\\\[5pt]
&\Rightarrow \quad - (1-\theta)(\frac{1}{2}(r - \frac{1}{2}\sigma^2)\frac{\Delta \tau}{\Delta x} + \frac{1}{2}\sigma^2\frac{\Delta \tau}{\Delta x^2})C_{j+1,k+1} + (1 + (1-\theta)(\sigma^2\frac{\Delta \tau}{\Delta x^2} + r\Delta \tau)C_{j,k+1} - (1-\theta)(\frac{1}{2}\sigma^2\frac{\Delta \tau}{\Delta x^2} - \frac{1}{2}(r - \frac{1}{2}\sigma^2 )\frac{\Delta \tau}{\Delta x})C_{j-1,k+1} \\\\[5pt]
&= -\theta(\frac{1}{2}(r - \frac{1}{2}\sigma^2)\frac{\Delta \tau}{\Delta x} + \frac{1}{2}\sigma^2\frac{\Delta \tau}{\Delta x^2})C_{j+1,k} + (1 - \theta(\sigma^2\frac{\Delta \tau}{\Delta x^2} + r\Delta \tau)C_{j,k} + \theta(\frac{1}{2}\sigma^2\frac{\Delta \tau}{\Delta x^2} - \frac{1}{2}(r - \frac{1}{2}\sigma^2 )\frac{\Delta \tau}{\Delta x})C_{j-1,k}, \\\\[5pt]
&\Rightarrow \quad \hat{l}_j C_{j-1,k+1} + \hat{c}_j C_{j,k+1} + \hat{u}_j C_{j+1,k+1} = l_j C_{j-1,k} + c_j C_{j,k} + u_j C_{j+1,k}
\end{align*}

其中，隐式格式部分：
\begin{align*}
\hat{l}_j &= - (1-\theta)(\frac{1}{2}\sigma^2\frac{\Delta \tau}{\Delta x^2} - \frac{1}{2}(r - \frac{1}{2}\sigma^2 )\frac{\Delta \tau}{\Delta x}), \\\\[5pt]
\hat{c}_j &= 1 + (1-\theta)(\sigma^2\frac{\Delta \tau}{\Delta x^2} + r\Delta \tau), \\\\[5pt]
\hat{u}_j &= - (1-\theta)(\frac{1}{2}(r - \frac{1}{2}\sigma^2)\frac{\Delta \tau}{\Delta x} + \frac{1}{2}\sigma^2\frac{\Delta \tau}{\Delta x^2})
\end{align*}

显示格式部分：
\begin{align*}
l_j &= \theta(\frac{1}{2}\sigma^2\frac{\Delta \tau}{\Delta x^2} - \frac{1}{2}(r - \frac{1}{2}\sigma^2 )\frac{\Delta \tau}{\Delta x}), \\\\[5pt]
c_j &= 1 -\theta(\sigma^2\frac{\Delta \tau}{\Delta x^2} + r\Delta \tau), \\\\[5pt]
u_j &= \theta(\frac{1}{2}(r - \frac{1}{2}\sigma^2)\frac{\Delta \tau}{\Delta x} + \frac{1}{2}\sigma^2\frac{\Delta \tau}{\Delta x^2})
\end{align*}


以上即为差分方程组。

同之前介绍的隐式格式，我们这里使用**Dirichlet**边界条件，则：

\begin{align*}
C_{0,k} = C(x_{min},\tau), \\\\[5pt]
C_{N,k} = C(x_{max},\tau)
\end{align*}

3.实现
---------------
***

这里我们略去了与之前一天相同的组件的定义：``BSMModel``, ``CallOption``。这些部分我们将其放入工具库：``lib.PDE``

```python
import scipy as sp
from scipy.linalg import solve_banded
from lib.PDE import BSMModel, CallOption
```

完整的Theta格式：``BSMThetaScheme``：

```python
class BSMThetaScheme:
    def __init__(self, model, payoff, T, M, N, theta):
        self.model = model
        self.T = T
        self.M = M
        self.N = N
        self.dt = self.T / self.M
        self.payoff = payoff
        self.x_min = model.x_min(self.T)
        self.x_max = model.x_max(self.T)
        self.dx = (self.x_max - self.x_min) / self.N
        self.C = np.zeros((self.N+1, self.M+1)) # 全部网格
        self.xArray = np.linspace(self.x_min, self.x_max, self.N+1)
        self.C[:,0] = map(self.payoff.ic, np.exp(self.xArray))
        
        sigma_square = self.model.sigma*self.model.sigma
        r = self.model.r
        self.theta = theta
        
        # Implicit 部分
        self.l_j_implicit = -(1.0 - self.theta) * (0.5*sigma_square*self.dt/self.dx/self.dx - 0.5 * (r - 0.5 * sigma_square)*self.dt/self.dx)
        self.c_j_implicit = 1.0 + (1.0 - self.theta) * (sigma_square*self.dt/self.dx/self.dx + r*self.dt)
        self.u_j_implicit = -(1.0 - self.theta) * (0.5*sigma_square*self.dt/self.dx/self.dx + 0.5 * (r - 0.5 * sigma_square)*self.dt/self.dx)
        
        # Explicit 部分
        self.l_j_explicit = self.theta * (0.5*sigma_square*self.dt/self.dx/self.dx - 0.5 * (r - 0.5 * sigma_square)*self.dt/self.dx)
        self.c_j_explicit = 1.0 - theta * (sigma_square*self.dt/self.dx/self.dx + r*self.dt)
        self.u_j_explicit = self.theta * (0.5*sigma_square*self.dt/self.dx/self.dx + 0.5 * (r - 0.5 * sigma_square)*self.dt/self.dx)
        
    def roll_back(self): 
        
        for k in range(0, self.M):
            rhs = self._apply_explicit(k)
            self._apply_implicti(k,rhs)
           
    def _apply_explicit(self, k):
        # 应用显示格式部分
        preSol =  np.copy(self.C[:,k])
        rhs = np.array([0.0] * (self.N-1))
            
        for i in range(self.N-1):
            rhs[i] = self.l_j_explicit * preSol[i] + self.c_j_explicit * preSol[i+1] + self.u_j_explicit * preSol[i+2]
        return rhs
    
    def _apply_implicti(self, k, rhs):
        # 应用隐式格式部分
        udiag = np.ones(self.N-1) * self.u_j_implicit
        ldiag =  np.ones(self.N-1) * self.l_j_implicit
        cdiag =  np.ones(self.N-1) * self.c_j_implicit
            
        mat = np.zeros((3,self.N-1))
        mat[0,:] = udiag
        mat[1,:] = cdiag
        mat[2,:] = ldiag
            
        # 应用左端边值条件
        v1 = self.payoff.bcl(math.exp(self.x_min), (k+1)*self.dt, self.model)
        rhs[0] -= self.l_j_implicit * v1
            
        # 应用右端边值条件
        v2 = self.payoff.bcr(math.exp(self.x_max), (k+1)*self.dt, self.model)
        rhs[-1] -= self.u_j_implicit * v2
            
        x = solve_banded((1,1), mat, rhs)
        self.C[1:self.N, k+1] = x
        self.C[0][k+1] = v1
        self.C[self.N][k+1] = v2
            
    def mesh_grids(self):
        tArray = np.linspace(0, self.T, self.M+1)
        tGrids, xGrids = np.meshgrid(tArray, self.xArray)
        return tGrids, xGrids
```

应用在一起：

```python
model = BSMModel(100.0, 0.05, 0.2)
payoff = CallOption(105.0)
scheme = BSMThetaScheme(model, payoff, 5.0, 100, 300, 0.5)
```

我们首先还是先获取解析解的结果：

```python
analyticPrice = BSMPrice(1, 105., 100., 0.05, 0.0, 0.2, 5.)
analyticPrice
```

我们固定时间方向网格数为500，分别计算不同$S$网格数情形下的结果。这里为了比较Theta格式与隐式格式的收敛速度，我们分别计算它们在不同网格下的结果:

```python
theta = 0.5 # Crank - Nicolson格式
timeStep = 500
xSteps = range(50,300,10)
crankNicolsonResult = []
implicitResult = []
explicitResult = []
for xStep in xSteps:
    model = BSMModel(100.0, 0.05, 0.2)
    payoff = CallOption(105.0)
    scheme = BSMThetaScheme(model, payoff, 5.0, timeStep, xStep, theta)
    scheme.roll_back()
    interp = CubicNaturalSpline(np.exp(scheme.xArray), scheme.C[:,-1])
    price = interp(100.0)
    crankNicolsonResult.append(price)
    
    scheme = BSMThetaScheme(model, payoff, 5.0, timeStep, xStep, 0.0)
    scheme.roll_back()
    interp = CubicNaturalSpline(np.exp(scheme.xArray), scheme.C[:,-1])
    price = interp(100.0)
    implicitResult.append(price)
```

我们可以画下收敛图：当$\theta=0.5$时候，即为Crank-Nicolson格式时，差分格式收敛素的显著高于隐式差分格式。

```python
anyRes = [analyticPrice['price'][1]] * len(xSteps)

pylab.figure(figsize = (16,8))
pylab.plot(xSteps, crankNicolsonResult, '-.', marker = 'o', markersize = 10)
pylab.plot(xSteps, implicitResult, '-.', marker = '^', markersize = 10)
pylab.plot(xSteps, anyRes, '--')
pylab.legend([u'Crank - Nicolson差分格式', u'隐式差分格式', u'解析解（欧式）' ], prop = font)
pylab.xlabel(u'价格方向网格步数', fontproperties = font)
pylab.title(u'Black - Scholes - Merton 有限差分法', fontproperties = font, fontsize = 20)
```
