当n很大，n个任意相互独立同分布（均值u、方差sigma都存在）的随机变量序列的平均值近似于正太分布  
N（u，sigma/sqrt（n））
n越大，平均值分布的标准差越小。
这也是赌场的生存法则，比如掷色子游戏，仅掷一次的风险远远大于掷多次，虽然两者的期望值是一样的。

```python
import numpy as np 
import scipy.stats as sps 
import matplotlib.pyplot as plt 
def mean(M,N):
	Xmn=sps.uniform.rvs(size=(M,N))	
	return np.mean(Xmn,axis=0)
plt.subplot(2,2,1)
plt.hist(mean(5,2000),bins=30)
plt.xlim(0.2,0.8)
plt.legend(['n=5'])

plt.subplot(2,2,2)
plt.hist(mean(50,2000),bins=30)
plt.xlim(0.2,0.8)
plt.legend(['n=50'])

plt.subplot(2,2,3)
plt.hist(mean(500,2000),bins=30)
plt.xlim(0.2,0.8)
plt.legend(['n=500'])

plt.subplot(2,2,4)
plt.hist(mean(5000,2000),bins=30)
plt.xlim(0.2,0.8)
plt.legend(['n=5000'])
plt.show()
```
