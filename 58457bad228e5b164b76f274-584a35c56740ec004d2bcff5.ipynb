### 用法： 股票收盘价涨停作为筛选条件，以涨停日为基准日期。
- unique_limit(beginDate='20150101',endDate='20161205',CT=20,ul/dl 涨跌幅度设置)
- 开始 结束 股票池持续交易时间（默认股票池为当前交易日前连续交易20个交易日）
- wave_analysis('20161103',lc=1)
- 交易日，lc=1表示当前交易日涨停，lc=n>1表示包括该交易日连续涨停n个交易日
- peried_trading_stocks_screen (idate) 
- 获得某交易日前 CT 个交易日内连续交易股票代码
- period_quotes_data(self,qstart,qend,lc=1):
- 筛选一段时间涨停板统计，lc表示涨停次数（连续）

```python
class unique_limit_up(object):
    def __init__(self,beginDate='20150101',endDate='20161202',CT=20,ul=1.096):
        self.beginDate= beginDate
        self.endDate= endDate
        self.universe= self.A_universe()
        self.date_list= self.date_list_get()
        self.CT= CT
        self.ul= ul
    
    def date_list_get(self):
        #交易日
        datedata=DataAPI.TradeCalGet(exchangeCD=u"XSHG",beginDate= self.beginDate, endDate= self.endDate, field=['calendarDate','isOpen']) 
        datedata=datedata[datedata['isOpen'] == 1]
        trading_date_list = map(lambda x: x[0:4]+x[5:7]+x[8:10], datedata['calendarDate'].values.tolist())    #多取一个月是为了判断是否在前二十个交易日上市
        return trading_date_list
    
    def A_universe(self):
        #股票池
        stocks = DataAPI.EquGet(equTypeCD='A',listStatusCD='L',field='secID,nonrestfloatA')
        universe = stocks['secID'].tolist()
        return universe
    
    def peried_trading_stocks_screen (self,idate):
        # 某交易日前20天内连续交易股票代码
        mch= self.date_list.index(idate)
        istart=self.date_list[mch-self.CT-1]
        iend=self.date_list[mch-1]
        
        tmp= DataAPI.MktEqudGet(secID=self.universe,ticker=u"",beginDate= istart, endDate= iend, isOpen="",field=['secID','isOpen'],pandas="1")
        tmp= tmp[tmp['isOpen']==1]
        secID_list=tmp['secID'].tolist()
        tsecID_list=sorted(list(set(secID_list)))
        
        c_list=[]
        for i in tsecID_list:
            isum= secID_list.count(i)
            if isum >= 20:
                c_list.append(i)
        return c_list
    
    def stock_date_list(self,idate):
        # 某交易日当日的涨停板 股票代码
        
        day_secID_list=self.peried_trading_stocks_screen(idate)
               
        tmp1= DataAPI.MktEqudGet(tradeDate=idate, secID=day_secID_list, isOpen="",
                                 field=['tradeDate','secID','acrPreClosePrice','openPrice','closePrice',
                                        'highestPrice','lowestPrice','turnoverValue','accumAdjFactor'],pandas="1").replace(0.0,np.nan).dropna()

        tmp1['openPrice']= tmp1['openPrice']/tmp1['accumAdjFactor']
        tmp1['highestPrice']= tmp1['highestPrice']/tmp1['accumAdjFactor']
        tmp1['closePrice']= tmp1['closePrice']/tmp1['accumAdjFactor']
        tmp1['lowestPrice']= tmp1['lowestPrice']/tmp1['accumAdjFactor']
        tmp1['up_limit']=tmp1['closePrice']/tmp1['openPrice']
        
        x_list= []
        for i in tmp1.index:
            if tmp1.ix[i]['up_limit']>=self.ul:
                x_list.append(i)
        tmp2= tmp1.ix[x_list]
        tmp2= tmp2['secID'].tolist()
        return tmp2
    def Four_special_plate(self,idate,lc=1):
        if lc==1:
            tmplist=self.stock_date_list(idate)
        elif lc>1:
            mch= self.date_list.index(idate)
            istart=self.date_list[mch+1-lc]
            
            datedata=DataAPI.TradeCalGet(exchangeCD=u"XSHG",beginDate= istart, endDate= idate, field=['calendarDate','isOpen']) 
            datedata=datedata[datedata['isOpen'] == 1]
            trading_date_list = map(lambda x: x[0:4]+x[5:7]+x[8:10], datedata['calendarDate'].values.tolist())    #多取一个月是为了判断是否在前二十个交易日上市
            tmplist=[]
            ttlist=[]
            for i in trading_date_list:
                tmp= self.stock_date_list(i)
                ttlist.extend(tmp)
            t1list=list(set(ttlist))
            for ic in t1list:
                if ttlist.count(ic)>=lc:
                    tmplist.append(ic)
        xtmplist=tmplist
        return xtmplist   
     
    def next_peried_quotes (self,idate,lc=1):
        # 某交易日后20日的报价单
        # 这里修改设置，包括涨停当日的股票报价，共21交易日数据
        mch= self.date_list.index(idate)
        istart=self.date_list[mch+1]
        iend=self.date_list[mch+20]
        #i_secID= self.stock_date_list(idate)
        i_secID=self.Four_special_plate(idate,lc=lc)
        
        tmp1= DataAPI.MktEqudGet(secID=i_secID, beginDate= idate, endDate=iend, isOpen="",
                         field=['tradeDate','secID','openPrice','closePrice',
                                'highestPrice','lowestPrice','turnoverValue','accumAdjFactor','negMarketValue'],pandas="1").replace(0.0,np.nan)
        
        tmp1['openPrice']= tmp1['openPrice']/tmp1['accumAdjFactor']
        tmp1['highestPrice']= tmp1['highestPrice']/tmp1['accumAdjFactor']
        tmp1['closePrice']= tmp1['closePrice']/tmp1['accumAdjFactor']
        tmp1['lowestPrice']= tmp1['lowestPrice']/tmp1['accumAdjFactor']
        tmp1['LogValue']= np.log(tmp1['negMarketValue'])
        return tmp1
    
    def wave_analysis(self,idate,lc=1):
        s1= self.next_peried_quotes(idate,lc=lc)
        t1= list(set(s1.secID.tolist()))
        t1= sorted(t1)
        #print'涨停板股票代码'
        #print t1
        print 'HC 20交易日最高价/涨停板收盘价， LC 20交易日最低价/涨停板收盘价， log 流通股对数市值'

        ts=pd.DataFrame()
        for isec in t1:
            t2= s1[s1['secID']== isec].sort_values('tradeDate')
            # 收盘价之比
            t2['CC1']= t2['closePrice'].shift(-1)/t2['closePrice']-1.0
            t2['CC3']= t2['closePrice'].shift(-3)/t2['closePrice']-1.0
            t2['CC5']= t2['closePrice'].shift(-5)/t2['closePrice']-1.0
            t2['CC10']= t2['closePrice'].shift(-10)/t2['closePrice']-1.0
            t2['CC15']= t2['closePrice'].shift(-15)/t2['closePrice']-1.0
            t2['CC20']= t2['closePrice'].shift(-20)/t2['closePrice']-1.0
            #日内涨幅
            t2['CO']= t2['closePrice']/ t2['openPrice'] -1.0
            #日内振幅
            t2['HL']= t2['highestPrice']/ t2['lowestPrice'] -1.0
            #跳开 明天开盘价比今天的收盘价
            t2['OC1']= t2['openPrice'].shift(-1)/t2['closePrice']-1.0
            t2['OC3']= t2['openPrice'].shift(-3)/t2['closePrice']-1.0
            t2['OC5']= t2['openPrice'].shift(-5)/t2['closePrice']-1.0
            t2['OC10']= t2['openPrice'].shift(-10)/t2['closePrice']-1.0
            t2['OC15']= t2['openPrice'].shift(-15)/t2['closePrice']-1.0
            t2['OC20']= t2['openPrice'].shift(-20)/t2['closePrice']-1.0
            t2['HC']= t2['highestPrice'].max()/t2['closePrice']-1.0
            t2['LC']= t2['lowestPrice'].min()/t2['closePrice']-1.0
            t3=t2.dropna()
            ts= ts.append(t3)
        ts=ts.fillna(0)
        ts= ts.set_index('secID')
        
        s1=ts
        
        sa1= s1.ix[:,['CC1','CC3','CC5','CC10','CC15','CC10']]
        sa2= sa1.T
        data= sa2
        index= np.arange(len(sa1))
        color=['r','g','b','c','m','y']
        fig= plt.figure(figsize=(len(sa1),14))
        ax1= fig.add_subplot(211)
        for i in range(len(sa2)):
            ax1.bar(index+0.125*i+0.125,sa2.iloc[i],width=0.125,color=color[i],alpha=1)
        xt=sa1.index.values.tolist()
        plt.xlim(0,len(sa1))
        ax1=plt.gca()
        ax1.set_xticks(np.arange(len(sa1)))
        ax1.set_xticklabels(xt,rotation=30,horizontalalignment='center')
        plt.title(u'渐次收盘价之比',fontproperties=font, fontsize=16)
        plt.grid()
        ax3=ax1.twinx()
        plt.plot(index+0.5,s1['HC'],'o', color='r')
        plt.plot(index+0.5,s1['LC'],'o', color='g')
        plt.legend(loc='best')

        sa3= s1.ix[:,['OC1','OC3','OC5','OC10','OC15','OC20']]
        sa4= sa3.T
        data= sa4
        index= np.arange(len(sa3))
        color=['r','g','b','c','m','y']
        fig= plt.figure(figsize=(len(sa3),14))
        ax2= fig.add_subplot(212)
        for i in range(len(sa4)):
            ax2.bar(index+0.125*i+0.125,sa4.iloc[i],width=0.125,color=color[i],alpha=1)
        xt=sa3.index.values.tolist()
        plt.xlim(-1,len(sa3)+1)
        ax2=plt.gca()
        ax2.set_xticks(np.arange(len(sa3)))
        ax2.set_xticklabels(xt,rotation=30,horizontalalignment='left')
        plt.title(u'渐次开盘价-收盘价之比',fontproperties=font, fontsize=16)
        plt.grid()
        ax4=ax2.twinx()
        plt.plot(index+0.5,s1['LogValue'],'o', color='b')
        plt.legend(loc='best')
        plt.show()
        
        s2=ts
        print '*************************************************************'
        print '日期',idate,'停板数量',len(s2)
        print '次日继续上涨数量',len(s2[s2.CC1>0]),'继续上涨比例',np.round((float(len(s2[s2.CC1>0]))/len(s2)),3)
        print '3日继续上涨数量',len(s2[s2.CC3>0]),'继续上涨比例',np.round((float(len(s2[s2.CC3>0]))/len(s2)),3)
        print '5日继续上涨数量',len(s2[s2.CC5>0]),'继续上涨比例',np.round((float(len(s2[s2.CC5>0]))/len(s2)),3)
        print '10日继续上涨数量',len(s2[s2.CC10>0]),'继续上涨比例',np.round((float(len(s2[s2.CC10>0]))/len(s2)),3)
        print '15日继续上涨数量',len(s2[s2.CC15>0]),'继续上涨比例',np.round((float(len(s2[s2.CC15>0]))/len(s2)),3)
        print '20日继续上涨数量',len(s2[s2.CC20>0]),'继续上涨比例',np.round((float(len(s2[s2.CC20>0]))/len(s2)),3)
        print''
        print '----------------------'
        print '追涨'
        print '次日高开数量',len(s2[s2.OC1>0]),'继续上涨比例',np.round((float(len(s2[s2.OC1>0]))/len(s2)),3)
        print '3日开盘/今日收盘价',len(s2[s2.OC3>0]),'继续上涨比例',np.round((float(len(s2[s2.OC3>0]))/len(s2)),3)
        print '5日开盘/今日收盘价',len(s2[s2.OC5>0]),'继续上涨比例',np.round((float(len(s2[s2.OC5>0]))/len(s2)),3)
        print '10日开盘/今日收盘价',len(s2[s2.OC10>0]),'继续上涨比例',np.round((float(len(s2[s2.OC10>0]))/len(s2)),3)
        print '15日开盘/今日收盘价',len(s2[s2.OC15>0]),'继续上涨比例',np.round((float(len(s2[s2.OC15>0]))/len(s2)),3)
        print '20日开盘/今日收盘价',len(s2[s2.OC20>0]),'继续上涨比例',np.round((float(len(s2[s2.OC20>0]))/len(s2)),3)
        print ''
        print ''
        print ''
        return ts   
    
    def statistics_wave_analysis(self,idate,lc=1):
        s1= self.next_peried_quotes(idate,lc=lc)
        t1= list(set(s1.secID.tolist()))
        ts=pd.DataFrame()
        for isec in t1:
            t2= s1[s1['secID']== isec].sort_values('tradeDate')
            # 收盘价之比
            t2['CC1']= t2['closePrice'].shift(-1)/t2['closePrice']-1.0
            t2['CC3']= t2['closePrice'].shift(-3)/t2['closePrice']-1.0
            t2['CC5']= t2['closePrice'].shift(-5)/t2['closePrice']-1.0
            t2['CC10']= t2['closePrice'].shift(-10)/t2['closePrice']-1.0
            t2['CC15']= t2['closePrice'].shift(-15)/t2['closePrice']-1.0
            t2['CC20']= t2['closePrice'].shift(-20)/t2['closePrice']-1.0
            #日内涨幅
            t2['CO']= t2['closePrice']/ t2['openPrice'] -1.0
            #日内振幅
            t2['HL']= t2['highestPrice']/ t2['lowestPrice'] -1.0
            #跳开 明天开盘价比今天的收盘价
            t2['OC1']= t2['openPrice'].shift(-1)/t2['closePrice']-1.0
            t2['OC3']= t2['openPrice'].shift(-3)/t2['closePrice']-1.0
            t2['OC5']= t2['openPrice'].shift(-5)/t2['closePrice']-1.0
            t2['OC10']= t2['openPrice'].shift(-10)/t2['closePrice']-1.0
            t2['OC15']= t2['openPrice'].shift(-15)/t2['closePrice']-1.0
            t2['OC20']= t2['openPrice'].shift(-20)/t2['closePrice']-1.0
            t2['HC']= t2['highestPrice'].max()/t2['closePrice']-1.0
            t2['LC']= t2['lowestPrice'].min()/t2['closePrice']-1.0
            t3=t2.dropna()
            ts= ts.append(t3)
        ts=ts.fillna(0)
        ts= ts.set_index('secID')
        return ts
    
    def period_quotes_data(self,qstart,qend,lc=1):
        #一段时间涨停板统计
        datedata=DataAPI.TradeCalGet(exchangeCD=u"XSHG",beginDate= qstart, endDate= qend, field=['calendarDate','isOpen']) 
        datedata=datedata[datedata['isOpen'] == 1]
        trading_date_list_t = map(lambda x: x[0:4]+x[5:7]+x[8:10], datedata['calendarDate'].values.tolist())    #多取一个月是为了判断是否在前二十个交易日上市
        quotes_tmp=pd.DataFrame()
        for ida in trading_date_list_t:
            #print type(ida)
            #tmp= self.statistics_wave_analysis(ida,lc=lc)
            try:
                tmp= self.statistics_wave_analysis(ida,lc=lc)
                quotes_tmp= quotes_tmp.append(tmp)
            except:
                pass
        return quotes_tmp
    def period_q_p(self,qstart,qend,lc=1):
        qtmp= self.period_quotes_data(qstart,qend,lc=lc)
        
```

```python
class unique_limit_down(object):
    def __init__(self,beginDate='20150101',endDate='20161202',CT=20,dl=0.92):
        self.beginDate= beginDate
        self.endDate= endDate
        self.universe= self.A_universe()
        self.date_list= self.date_list_get()
        self.CT= CT
        self.dl= dl
    
    def date_list_get(self):
        #交易日
        datedata=DataAPI.TradeCalGet(exchangeCD=u"XSHG",beginDate= self.beginDate, endDate= self.endDate, field=['calendarDate','isOpen']) 
        datedata=datedata[datedata['isOpen'] == 1]
        trading_date_list = map(lambda x: x[0:4]+x[5:7]+x[8:10], datedata['calendarDate'].values.tolist())    #多取一个月是为了判断是否在前二十个交易日上市
        return trading_date_list
    
    def A_universe(self):
        #股票池
        stocks = DataAPI.EquGet(equTypeCD='A',listStatusCD='L',field='secID,nonrestfloatA')
        universe = stocks['secID'].tolist()
        return universe
    
    def peried_trading_stocks_screen (self,idate):
        # 某交易日前20天内连续交易股票代码
        mch= self.date_list.index(idate)
        istart=self.date_list[mch-self.CT-1]
        iend=self.date_list[mch-1]
        
        tmp= DataAPI.MktEqudGet(secID=self.universe,ticker=u"",beginDate= istart, endDate= iend, isOpen="",field=['secID','isOpen'],pandas="1")
        tmp= tmp[tmp['isOpen']==1]
        secID_list=tmp['secID'].tolist()
        tsecID_list=sorted(list(set(secID_list)))
        
        c_list=[]
        for i in tsecID_list:
            isum= secID_list.count(i)
            if isum >= 20:
                c_list.append(i)
        return c_list
    
    def stock_date_list(self,idate):
        # 某交易日当日的涨停板 股票代码
        
        day_secID_list=self.peried_trading_stocks_screen(idate)
               
        tmp1= DataAPI.MktEqudGet(tradeDate=idate, secID=day_secID_list, isOpen="",
                                 field=['tradeDate','secID','acrPreClosePrice','openPrice','closePrice',
                                        'highestPrice','lowestPrice','turnoverValue','accumAdjFactor'],pandas="1").replace(0.0,np.nan).dropna()

        tmp1['openPrice']= tmp1['openPrice']/tmp1['accumAdjFactor']
        tmp1['highestPrice']= tmp1['highestPrice']/tmp1['accumAdjFactor']
        tmp1['closePrice']= tmp1['closePrice']/tmp1['accumAdjFactor']
        tmp1['lowestPrice']= tmp1['lowestPrice']/tmp1['accumAdjFactor']
        tmp1['up_limit']=tmp1['closePrice']/tmp1['openPrice']
        
        x_list= []
        for i in tmp1.index:
            if tmp1.ix[i]['up_limit']<self.dl:
                x_list.append(i)
        tmp2= tmp1.ix[x_list]
        tmp2= tmp2['secID'].tolist()
        return tmp2
    def Four_special_plate(self,idate,lc=1):
        if lc==1:
            tmplist=self.stock_date_list(idate)
        elif lc>1:
            mch= self.date_list.index(idate)
            istart=self.date_list[mch+1-lc]
            
            datedata=DataAPI.TradeCalGet(exchangeCD=u"XSHG",beginDate= istart, endDate= idate, field=['calendarDate','isOpen']) 
            datedata=datedata[datedata['isOpen'] == 1]
            trading_date_list = map(lambda x: x[0:4]+x[5:7]+x[8:10], datedata['calendarDate'].values.tolist())    #多取一个月是为了判断是否在前二十个交易日上市
            tmplist=[]
            ttlist=[]
            for i in trading_date_list:
                tmp= self.stock_date_list(i)
                ttlist.extend(tmp)
            t1list=list(set(ttlist))
            for ic in t1list:
                if ttlist.count(ic)>=lc:
                    tmplist.append(ic)
        xtmplist=tmplist
        return xtmplist   
     
    def next_peried_quotes (self,idate,lc=1):
        # 某交易日后20日的报价单
        # 这里修改设置，包括涨停当日的股票报价，共21交易日数据
        mch= self.date_list.index(idate)
        istart=self.date_list[mch+1]
        iend=self.date_list[mch+20]
        #i_secID= self.stock_date_list(idate)
        i_secID=self.Four_special_plate(idate,lc=lc)
        
        tmp1= DataAPI.MktEqudGet(secID=i_secID, beginDate= idate, endDate=iend, isOpen="",
                         field=['tradeDate','secID','openPrice','closePrice',
                                'highestPrice','lowestPrice','turnoverValue','accumAdjFactor','negMarketValue'],pandas="1").replace(0.0,np.nan)
        
        tmp1['openPrice']= tmp1['openPrice']/tmp1['accumAdjFactor']
        tmp1['highestPrice']= tmp1['highestPrice']/tmp1['accumAdjFactor']
        tmp1['closePrice']= tmp1['closePrice']/tmp1['accumAdjFactor']
        tmp1['lowestPrice']= tmp1['lowestPrice']/tmp1['accumAdjFactor']
        tmp1['LogValue']= np.log(tmp1['negMarketValue'])
        return tmp1
    
    def wave_analysis(self,idate,lc=1):
        s1= self.next_peried_quotes(idate,lc=lc)
        t1= list(set(s1.secID.tolist()))
        t1= sorted(t1)
        #print'涨停板股票代码'
        #print t1
        print 'HC 20交易日最高价/涨停板收盘价， LC 20交易日最低价/涨停板收盘价， log 流通股对数市值'

        ts=pd.DataFrame()
        for isec in t1:
            t2= s1[s1['secID']== isec].sort_values('tradeDate')
            # 收盘价之比
            t2['CC1']= t2['closePrice'].shift(-1)/t2['closePrice']-1.0
            t2['CC3']= t2['closePrice'].shift(-3)/t2['closePrice']-1.0
            t2['CC5']= t2['closePrice'].shift(-5)/t2['closePrice']-1.0
            t2['CC10']= t2['closePrice'].shift(-10)/t2['closePrice']-1.0
            t2['CC15']= t2['closePrice'].shift(-15)/t2['closePrice']-1.0
            t2['CC20']= t2['closePrice'].shift(-20)/t2['closePrice']-1.0
            #日内涨幅
            t2['CO']= t2['closePrice']/ t2['openPrice'] -1.0
            #日内振幅
            t2['HL']= t2['highestPrice']/ t2['lowestPrice'] -1.0
            #跳开 明天开盘价比今天的收盘价
            t2['OC1']= t2['openPrice'].shift(-1)/t2['closePrice']-1.0
            t2['OC3']= t2['openPrice'].shift(-3)/t2['closePrice']-1.0
            t2['OC5']= t2['openPrice'].shift(-5)/t2['closePrice']-1.0
            t2['OC10']= t2['openPrice'].shift(-10)/t2['closePrice']-1.0
            t2['OC15']= t2['openPrice'].shift(-15)/t2['closePrice']-1.0
            t2['OC20']= t2['openPrice'].shift(-20)/t2['closePrice']-1.0
            t2['HC']= t2['highestPrice'].max()/t2['closePrice']-1.0
            t2['LC']= t2['lowestPrice'].min()/t2['closePrice']-1.0
            t3=t2.dropna()
            ts= ts.append(t3)
        ts=ts.fillna(0)
        ts= ts.set_index('secID')
        
        s1=ts
        
        sa1= s1.ix[:,['CC1','CC3','CC5','CC10','CC15','CC10']]
        sa2= sa1.T
        data= sa2
        index= np.arange(len(sa1))
        color=['r','g','b','c','m','y']
        fig= plt.figure(figsize=(len(sa1),14))
        ax1= fig.add_subplot(211)
        for i in range(len(sa2)):
            ax1.bar(index+0.125*i+0.125,sa2.iloc[i],width=0.125,color=color[i],alpha=1)
        xt=sa1.index.values.tolist()
        plt.xlim(0,len(sa1))
        ax1=plt.gca()
        ax1.set_xticks(np.arange(len(sa1)))
        ax1.set_xticklabels(xt,rotation=30,horizontalalignment='center')
        plt.title(u'渐次收盘价之比',fontproperties=font, fontsize=16)
        plt.grid()
        ax3=ax1.twinx()
        plt.plot(index+0.5,s1['HC'],'o', color='r')
        plt.plot(index+0.5,s1['LC'],'o', color='g')
        plt.legend(loc='best')

        sa3= s1.ix[:,['OC1','OC3','OC5','OC10','OC15','OC20']]
        sa4= sa3.T
        data= sa4
        index= np.arange(len(sa3))
        color=['r','g','b','c','m','y']
        fig= plt.figure(figsize=(len(sa3),14))
        ax2= fig.add_subplot(212)
        for i in range(len(sa4)):
            ax2.bar(index+0.125*i+0.125,sa4.iloc[i],width=0.125,color=color[i],alpha=1)
        xt=sa3.index.values.tolist()
        plt.xlim(-1,len(sa3)+1)
        ax2=plt.gca()
        ax2.set_xticks(np.arange(len(sa3)))
        ax2.set_xticklabels(xt,rotation=30,horizontalalignment='left')
        plt.title(u'渐次开盘价-收盘价之比',fontproperties=font, fontsize=16)
        plt.grid()
        ax4=ax2.twinx()
        plt.plot(index+0.5,s1['LogValue'],'o', color='b')
        plt.legend(loc='best')
        plt.show()
        
        s2=ts
        print '*************************************************************'
        print '日期',idate,'停板数量',len(s2)
        print '次日继续下跌数量',len(s2[s2.CC1<0]),'继续下跌比例',np.round((float(len(s2[s2.CC1<0]))/len(s2)),3)
        print '3日继续下跌数量',len(s2[s2.CC3<0]),'继续下跌比例',np.round((float(len(s2[s2.CC3<0]))/len(s2)),3)
        print '5日继续下跌数量',len(s2[s2.CC5<0]),'继续下跌比例',np.round((float(len(s2[s2.CC5<0]))/len(s2)),3)
        print '10日继下跌涨数量',len(s2[s2.CC10>0]),'继续下跌比例',np.round((float(len(s2[s2.CC10<0]))/len(s2)),3)
        print '15日继下跌涨数量',len(s2[s2.CC15<0]),'继续下跌比例',np.round((float(len(s2[s2.CC15<0]))/len(s2)),3)
        print '20日继下跌涨数量',len(s2[s2.CC20<0]),'继续下跌比例',np.round((float(len(s2[s2.CC20<0]))/len(s2)),3)
        print''
        print '----------------------'
        print '杀跌'
        print '次日低开数量',len(s2[s2.OC1<0]),'继续下跌比例',np.round((float(len(s2[s2.OC1<0]))/len(s2)),3)
        print '3日开盘/今日收盘价',len(s2[s2.OC3<0]),'继续下跌比例',np.round((float(len(s2[s2.OC3<0]))/len(s2)),3)
        print '5日开盘/今日收盘价',len(s2[s2.OC5<0]),'继续下跌比例',np.round((float(len(s2[s2.OC5<0]))/len(s2)),3)
        print '10日开盘/今日收盘价',len(s2[s2.OC10<0]),'继续下跌比例',np.round((float(len(s2[s2.OC10<0]))/len(s2)),3)
        print '15日开盘/今日收盘价',len(s2[s2.OC15<0]),'继续下跌比例',np.round((float(len(s2[s2.OC15<0]))/len(s2)),3)
        print '20日开盘/今日收盘价',len(s2[s2.OC20<0]),'继续下跌比例',np.round((float(len(s2[s2.OC20<0]))/len(s2)),3)
        print ''
        print ''
        print ''
        return ts   
    
    def statistics_wave_analysis(self,idate,lc=1):
        s1= self.next_peried_quotes(idate,lc=lc)
        t1= list(set(s1.secID.tolist()))
        ts=pd.DataFrame()
        for isec in t1:
            t2= s1[s1['secID']== isec].sort_values('tradeDate')
            # 收盘价之比
            t2['CC1']= t2['closePrice'].shift(-1)/t2['closePrice']-1.0
            t2['CC3']= t2['closePrice'].shift(-3)/t2['closePrice']-1.0
            t2['CC5']= t2['closePrice'].shift(-5)/t2['closePrice']-1.0
            t2['CC10']= t2['closePrice'].shift(-10)/t2['closePrice']-1.0
            t2['CC15']= t2['closePrice'].shift(-15)/t2['closePrice']-1.0
            t2['CC20']= t2['closePrice'].shift(-20)/t2['closePrice']-1.0
            #日内涨幅
            t2['CO']= t2['closePrice']/ t2['openPrice'] -1.0
            #日内振幅
            t2['HL']= t2['highestPrice']/ t2['lowestPrice'] -1.0
            #跳开 明天开盘价比今天的收盘价
            t2['OC1']= t2['openPrice'].shift(-1)/t2['closePrice']-1.0
            t2['OC3']= t2['openPrice'].shift(-3)/t2['closePrice']-1.0
            t2['OC5']= t2['openPrice'].shift(-5)/t2['closePrice']-1.0
            t2['OC10']= t2['openPrice'].shift(-10)/t2['closePrice']-1.0
            t2['OC15']= t2['openPrice'].shift(-15)/t2['closePrice']-1.0
            t2['OC20']= t2['openPrice'].shift(-20)/t2['closePrice']-1.0
            t2['HC']= t2['highestPrice'].max()/t2['closePrice']-1.0
            t2['LC']= t2['lowestPrice'].min()/t2['closePrice']-1.0
            t3=t2.dropna()
            ts= ts.append(t3)
        ts=ts.fillna(0)
        ts= ts.set_index('secID')
        return ts
    
    def period_quotes_data(self,qstart,qend,lc=1):
        #一段时间涨停板统计
        datedata=DataAPI.TradeCalGet(exchangeCD=u"XSHG",beginDate= qstart, endDate= qend, field=['calendarDate','isOpen']) 
        datedata=datedata[datedata['isOpen'] == 1]
        trading_date_list_t = map(lambda x: x[0:4]+x[5:7]+x[8:10], datedata['calendarDate'].values.tolist())    #多取一个月是为了判断是否在前二十个交易日上市
        quotes_tmp=pd.DataFrame()
        for ida in trading_date_list_t:
            #print type(ida)
            #tmp= self.statistics_wave_analysis(ida,lc=lc)
            try:
                tmp= self.statistics_wave_analysis(ida,lc=lc)
                quotes_tmp= quotes_tmp.append(tmp)
            except:
                pass
        return quotes_tmp
    def period_q_p(self,qstart,qend,lc=1):
        qtmp= self.period_quotes_data(qstart,qend,lc=lc)
        
```
