### 从wind导出的期权数据与对应标的的50ETF标的数据长度不一致
### 通过调整，根据50ETF价格选出当天收盘价距离执行价格最近的合约

```python
import pandas as pd

# data1 = pd.read_excel('optdata.xlsx')
# data2 = pd.read_excel('50etf.xlsx')
#
# opt_data1 = pd.merge(data1,data2)                                    将每天50ETF的价格填入每天的期权合约后面

data = pd.read_excel('opt_data1.xlsx')                               # 读取文件
df1 = pd.DataFrame(data, columns=['date', 'opt_abbr', 'k', '50ETF']) # 定义表格中的各个column
df1['diff'] = abs(df1['k'] - df1['50ETF'])                           # 求执行价和50ETF之间的最小值
df2 = df1.set_index('opt_abbr')                                      # 将index改为期权合约简称
df2.groupby('date').agg({'diff':'min'})                              # 根据日期groupby，并选出对应日期下最小的diff
df3 = df2.groupby('date')['diff'].apply(lambda d: d.argmin())        # 选出当天'diff'最小的合约，并返回日期和合约简称

df3.to_excel('opt1.xlsx')
```
