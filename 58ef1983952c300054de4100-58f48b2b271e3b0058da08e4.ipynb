```python
# step 1 : Load Macro Data
def myMacroDataload():
    # 1、基本参数设置
    mystart = '2007-04-01'
    myend = '2016-12-31'
    myfield = 'periodDate,dataValue'
    data = pd.DataFrame() # month
    # 2、获取原始宏观和市场数据
    # (1)：美元兑人民币中间价（日）
    # startDate : 2006-01-04
    fx = DataAPI.ChinaDataExchangeRateGet(indicID=u"M110000003",beginDate=mystart,endDate=myend,field=myfield) 
    # 因为取到数据的频度不同，需要进行数据变频操作，即将日度数据转化为月度，以下日度数据均采用此操作
    # 首先利用CAL.PyCAL模块转换年份月数格式的日期，即yyyymm
    # data type convert to that PyCAL support
    fx['periodDate'] = fx['periodDate'].apply(cpc.Date.parseISO)
    # generate 'yyyymm' series
    fx['yyyymm'] = fx['periodDate'].apply(cpc.Date.month)+fx['periodDate'].apply(cpc.Date.year)*100
    # fx freq chg by groupby method
    fx = fx.groupby('yyyymm').head(1) #时间倒序
    # use np.array() to keep right index
    data['fx'] = np.array(fx.dataValue)
    data['yyyymm']=np.array(fx['yyyymm'])
    del fx # 随时删除无用数据
    # (2)：M0 和 M2（月）
    # Uqer提取出来是货币供应量M0和M2数据，前半部分是供应量，后半部分是同比增速，按照日期由近及远倒序排列
    # 参考申万的研究，我们选择后面的增速部分
    m0 = DataAPI.ChinaDataMoneyStatisticsGet(indicName=u"M0",beginDate=mystart,endDate=myend,field=myfield)  #时间倒序
    n = m0['dataValue'].shape[0]
    data['m0'] = np.array(m0.dataValue[int(n/2):]) # 需要转化为array，否则index会乱掉
    data['periodDate'] = np.array(m0['periodDate'][int(n/2):])
    m2 = DataAPI.ChinaDataMoneyStatisticsGet(indicName=u"M2",beginDate=mystart,endDate=myend,field=myfield) #时间倒序
    n = m2['dataValue'].shape[0]
    data['m2'] = np.array(m2.dataValue[int(n/2):])
    del m0,m2
    # (3): CPI(月)
    cpi = DataAPI.ChinaDataCPIGet(indicID=u"M030000003",beginDate=mystart,endDate=myend,field=myfield) #时间倒序
    data['cpi'] = cpi.dataValue
    # 调整data中日期的顺序
    data.index = data['yyyymm'] 
    data.sort_index(inplace=True)
    del cpi
    # (4):中证1000PE - 沪深3000PE（日）
    # 日度数据，需要变频；正常顺序，不需设置index并sort，直接merge
    # Uqer没有中证1000估值数据，采用中小板综指（399101）代替
    # a. large cap PE
    # CSI300 PE start Date : 2007-04-10
    # 根据PEType数据，取对应的TTM数据
    # PE类型，1--市盈率ttm；2--动态市盈率
    # type1类PE很多数据为NaN，采用type2    
    largeCapPE = DataAPI.MktIdxdEvalGet(ticker=u'000300',beginDate=mystart,endDate=myend,field= ['ticker','tradeDate','PEValue','PEType']) #时间顺序
    largeCapPE= largeCapPE[largeCapPE['PEType']==2] 
    largeCapPE.fillna(method='bfill',inplace=True)
    # freq chg
    largeCapPE['tradeDate'] = largeCapPE['tradeDate'].apply(cpc.Date.parseISO)  
    largeCapPE['yyyymm'] = largeCapPE['tradeDate'].apply(cpc.Date.month)+largeCapPE['tradeDate'].apply(cpc.Date.year)*100
    largeCapPE = largeCapPE.groupby('yyyymm').head(1)
    largeCapPE['PElargecap'] = largeCapPE.pop('PEValue')
    largeCapPE.pop('ticker')
    largeCapPE.pop('PEType')
    largeCapPE.pop('tradeDate')
    # merge data
    data = pd.merge(data,largeCapPE,on='yyyymm') # merge之后会丢失index
    # large cap PE data在2016年12月以后的数据可能有问题，wind显示PE_ttm在13X左右，而这部分提取到的超过了100，看是否为数据库的问题，还是提取用到的api函数错了######
    # b. small cap PE
    smallCapPE = DataAPI.MktIdxdEvalGet(ticker=u'399101',beginDate=mystart,endDate=myend,field= ['ticker','tradeDate','PEValue','PEType']) 
    smallCapPE = smallCapPE[smallCapPE['PEType']==2]    
    smallCapPE.fillna(method='bfill',inplace=True)
    smallCapPE['tradeDate'] = smallCapPE['tradeDate'].apply(cpc.Date.parseISO)
    smallCapPE['yyyymm'] = smallCapPE['tradeDate'].apply(cpc.Date.month)+ smallCapPE['tradeDate'].apply(cpc.Date.year)*100
    smallCapPE = smallCapPE.groupby('yyyymm').head(1)
    smallCapPE['PEsmallcap'] = smallCapPE.pop('PEValue')
    smallCapPE.pop('ticker')
    smallCapPE.pop('PEType') 
    smallCapPE.pop('tradeDate')
    data = pd.merge(data,smallCapPE,on='yyyymm')
    data['pe_smb'] = data['PEsmallcap'] - data['PElargecap']
    del largeCapPE,smallCapPE
    # (5):中证1000PB - 沪深3000PB（日） 
    # Uqer没有中证1000估值数据，采用中小板综指（399101）代替### 该部分数据可能有问题
    largeCapPB = DataAPI.MktIdxdEvalGet(ticker=u'000300',beginDate=mystart,endDate=myend,field= ['tradeDate','PB']) 
    largeCapPB.fillna(method='bfill',inplace=True)
    largeCapPB['tradeDate'] = largeCapPB['tradeDate'].apply(cpc.Date.parseISO)
    largeCapPB['yyyymm'] = largeCapPB['tradeDate'].apply(cpc.Date.month)+ largeCapPB['tradeDate'].apply(cpc.Date.year)*100
    largeCapPB = largeCapPB.groupby('yyyymm').head(1)
    largeCapPB['PBlargecap'] = largeCapPB.pop('PB')
    largeCapPB.pop('tradeDate')
    data = pd.merge(data,largeCapPB,on='yyyymm')
    smallCapPB = DataAPI.MktIdxdEvalGet(ticker=u'399101',beginDate=mystart,endDate=myend,field= ['tradeDate','PB']) 
    smallCapPB.fillna(method='bfill',inplace=True)
    smallCapPB['tradeDate'] = smallCapPB['tradeDate'].apply(cpc.Date.parseISO)
    smallCapPB['yyyymm'] = smallCapPB['tradeDate'].apply(cpc.Date.month)+ smallCapPB['tradeDate'].apply(cpc.Date.year)*100
    smallCapPB = smallCapPB.groupby('yyyymm').head(1)
    smallCapPB['PBsmallcap'] = smallCapPB.pop('PB')
    smallCapPB.pop('tradeDate')
    data = pd.merge(data,smallCapPB,on='yyyymm')
    data['pb_smb'] = data['PBsmallcap'] - data['PBlargecap']
    del largeCapPB,smallCapPB
    # （6）:中证全指ret - 中证全债ret（月）
    # Uqer里面中证全指数（000985.CSI）据起始于2011年8月2日，无法满足要求
    # 采用上证综指（000001.SH）代替
    myPreviousYear = str(int(mystart[:4])-1)+mystart[4:] # 1 year 
    csiAllStockret = DataAPI.MktIdxmGet(ticker=u'000001',beginDate=myPreviousYear,endDate=myend,field='endDate,chgPct') #时间倒序
    csiAllBondret = DataAPI.MktIdxmGet(ticker=u'h11001',beginDate=myPreviousYear,endDate=myend,field='endDate,chgPct') #时间倒序
    csiAllBSret = pd.DataFrame(np.array(csiAllStockret['chgPct']-csiAllBondret['chgPct']),columns=['csiAllBSretgap']) 
    csiAllStockret['endDate'] = csiAllStockret['endDate'].apply(cpc.Date.parseISO)
    csiAllBSret['yyyymm'] = csiAllStockret['endDate'].apply(cpc.Date.month)+csiAllStockret['endDate'].apply(cpc.Date.year)*100
    csiAllBSret.index = csiAllBSret['yyyymm'] 
    csiAllBSret.sort_index(inplace=True)
    data = pd.merge(data,csiAllBSret,on='yyyymm')
    del csiAllStockret,csiAllBondret,csiAllBSret
    # （7）：全A股息率-10年期国债到期收益率
    # pass
    # （8）：3M期国债YTM环比变化率
    # pass
    # （9）：中证1000指数histvol - 沪深300指数histvol (日)
    # 将每日收益累加为月度收益，计算月波动率，转化为年波动率
    myPreviousYear = str(int(mystart[:4])-1)+mystart[4:] # 1 year
    # small cap volatility
    csi1000Stockret = DataAPI.MktIdxdGet(ticker=u'000852',beginDate=myPreviousYear,endDate=myend,field='tradeDate,CHGPct') # 时间顺序，Daily data
    csi1000Stockret['tradeDate'] = csi1000Stockret['tradeDate'].apply(cpc.Date.parseISO) # date type ISO
    csi1000Stockret['yyyymm'] = csi1000Stockret['tradeDate'].apply(cpc.Date.year)*100+csi1000Stockret['tradeDate'].apply(cpc.Date.month)# generate year-month series
    smallcapvol = pd.DataFrame(csi1000Stockret.groupby('yyyymm')['CHGPct'].sum()) # sum daily return to monthly return
    smallcapvol['smallcapvol'] = pd.rolling_std(smallcapvol['CHGPct'],12)*np.sqrt(12) # month vol to year
    smallcapvol.pop('CHGPct')
    smallcapvol.dropna(axis=0,inplace=True) # del nan value in row
    smallcapvol['yyyymm'] = smallcapvol.index
    data = pd.merge(data,smallcapvol,on='yyyymm',how='inner') #因vol计算时向前取了一年的数据，两者的yyyymm不太一致，以data为准
    # large cap volatility
    csi300Stockret = DataAPI.MktIdxdGet(ticker=u'000300',beginDate=myPreviousYear,endDate=myend,field='tradeDate,CHGPct')# 时间顺序，Daily data
    csi300Stockret['tradeDate'] = csi300Stockret['tradeDate'].apply(cpc.Date.parseISO)
    csi300Stockret['yyyymm'] = csi300Stockret['tradeDate'].apply(cpc.Date.month) + csi300Stockret['tradeDate'].apply(cpc.Date.year)*100
    largecapvol = pd.DataFrame(csi300Stockret.groupby('yyyymm').sum())
    largecapvol['largecapvol'] = pd.rolling_std(largecapvol['CHGPct'],12) * np.sqrt(12)
    largecapvol.pop('CHGPct')
    largecapvol.dropna(axis=0,inplace=True)
    largecapvol['yyyymm'] = largecapvol.index
    data = pd.merge(data,largecapvol,on='yyyymm',how='inner')
    data['vol_smb'] = data['smallcapvol'] - data['largecapvol']
    del csi1000Stockret,csi300Stockret,smallcapvol,largecapvol
    # （10）：中证1000指数ret - 沪深300指数ret（月）
    # 中证1000指数月收益
    csi1000StockMret = DataAPI.MktIdxmGet(ticker=u'000852',beginDate=mystart,endDate=myend,field='endDate,chgPct')  # 时间倒序
    csi1000StockMret['endDate'] = csi1000StockMret['endDate'].apply(cpc.Date.parseISO)
    csi1000StockMret['yyyymm'] = csi1000StockMret['endDate'].apply(cpc.Date.month)+csi1000StockMret['endDate'].apply(cpc.Date.year)*100
    csi1000StockMret.index = csi1000StockMret['yyyymm']
    csi1000StockMret.sort_index(inplace=True)
    # 沪深300指数月收益
    csi300StockMret = DataAPI.MktIdxmGet(ticker=u'000300',beginDate=mystart,endDate=myend,field='endDate,chgPct') # 时间倒序
    csi300StockMret['endDate'] = csi300StockMret['endDate'].apply(cpc.Date.parseISO)
    csi300StockMret['yyyymm'] = csi300StockMret['endDate'].apply(cpc.Date.month)+csi300StockMret['endDate'].apply(cpc.Date.year)*100
    csi300StockMret.index = csi300StockMret['yyyymm']
    csi300StockMret.sort_index(inplace=True)
    data['mret_smb'] = np.array(csi1000StockMret['chgPct']) - np.array(csi300StockMret['chgPct'])
    del csi1000StockMret,csi300StockMret
    # （11） 行业走势度分化率
    # pass
    # 3. set index
    data.index = data.pop('yyyymm')
    return data

#rawdata
```

```python
# step 2 : Define normalize function
def myNormalizeFun(seriesdata,method='normal'):
    if method == 'normal':
        seriesdata[seriesdata>np.percentile(seriesdata,99)] = np.percentile(seriesdata,99)
        seriesdata[seriesdata<np.percentile(seriesdata,1)] = np.percentile(seriesdata,1)
    elif method == 'MedCouple':
        # MedCouple Approach
        # 25% pct and 75% pct
        Q1 = np.percentile(seriesdata,25)
        Q3 = np.percentile(seriesdata,75)
        IQR = Q3-Q1
        # calculate median
        md = np.median(seriesdata)
        # cal mc
        xi = seriesdata[seriesdata > md]
        xj = seriesdata[seriesdata < md]
        x = ((xi-md)-(md-xj))/(xi-xj)
        mc = np.median(x)
        # cal U and L
        if mc >= 0:
            L = Q1 - 1.5*np.exp(-3.5*mc)* IQR
            U = Q3 + 1.5*np.exp(4*mc)* IQR
        else:
            L = Q1 - 1.5*np.exp(-4*mc)* IQR
            U = Q3 + 1.5*np.exp(3.5*mc)* IQR
        seriesdata[seriesdata<L] = L
        seriesdata[seriesdata>U] = U
    datamedian = np.median(seriesdata)
    datastd = np.nanstd(seriesdata)
    dataNornamlized = (seriesdata-datamedian)/datastd 
    return dataNornamlized
```

```python
# step 3 : 生成虚拟变量
# data : 117 rows × 16 columns
# 大小盘风格差的虚拟变量，即 IF(（小盘月收益率-大盘月收益率）>0)
def myDummyVarGener(rawdata):
    Y = np.zeros_like(rawdata['mret_smb'])
    Y[np.where(rawdata['mret_smb']>0)]=1
    return Y
```

```python
# step 4：生成标准化自变量
def myGenerFitdata(xraw):
    # 生成所需因变量数据，分为xold和xnew
    # xold用于回归生成因变量系数
    # xnew 
    x = pd.DataFrame()
    xold = pd.DataFrame() # previous data for fit
    xnew = pd.DataFrame()
    x['fx'] = myNormalizeFun(xraw['fx'].values) 
    xold['fx'] = x['fx'].shift(1).fillna(method='bfill') # fill nan by next data
    xnew['fx'] = [x['fx'].iloc[-1]]
    x['m0'] = myNormalizeFun(xraw['m0'].values)
    xold['m0'] = x['m0'].shift(2).fillna(method='bfill')
    xnew['m0'] = [x['m0'].iloc[-2]]
    x['m2'] = myNormalizeFun(xraw['m2'].values)
    xold['m2'] = x['m2'].shift(2).fillna(method='bfill')
    xnew['m2']= [x['m2'].iloc[-2]]
    x['pe_smb'] = myNormalizeFun(xraw['pe_smb'].values)
    xold['pe_smb'] = x['pe_smb'].shift(1).fillna(method='bfill')
    xnew['pe_smb'] = [x['pe_smb'].iloc[-1]]
    x['pb_smb'] = myNormalizeFun(xraw['pb_smb'].values)
    xold['pb_smb'] = x['pb_smb'].shift(1).fillna(method='bfill')
    xnew['pb_smb'] = [x['pb_smb'].iloc[-1]]
    x['retgap0'] = myNormalizeFun(xraw['csiAllBSretgap'].values)
    xold['retgap0'] = x['retgap0'].shift(1).fillna(method='bfill')
    xnew['retgap0'] = [x['retgap0'].iloc[-1]]
    x['retgap1'] = myNormalizeFun(xraw['csiAllBSretgap'].values)
    xold['retgap1'] = x['retgap1'].shift(1).fillna(method='bfill')
    xnew['retgap1'] = [x['retgap1'].iloc[-1]]
    x['vol_smb'] = myNormalizeFun(xraw['vol_smb'].values)    
    xold['vol_smb'] = x['vol_smb'].shift(1).fillna(method='bfill')
    xnew['vol_smb'] = [x['vol_smb'].iloc[-1]]
    return xold,xnew
```

```python
# step 5 : 利用Logtistic回归以滚动时间窗口进行预测
# 回归时要考虑数据的滞后期问题，将宏观数据滞后1~2期与序列进行回归
# iloc:integer-location based indexing for selection by position
# 这个部分没有考虑到不显著的变量
# 待解决问题：样本内预测准确率如何确定？
rawdata = myMacroDataload()
#print rawdata.head(15)
Y = myDummyVarGener(rawdata)
datewindow = 20 # 申万研究推荐采用20日作为滚动预测的时间窗口
predictResList = []
ttick = []
for t in range(0,len(Y)-datewindow): # implied t: start month to end-window month
    xraw = rawdata.iloc[t:t+datewindow+1] # iloc:按行取数据；t要+1，否则最后一个值取不到,length=window
    ttick.append(rawdata.index[t+datewindow]) # get current Date
    xold,xnew = myGenerFitdata(xraw)
    y = Y[t:t+datewindow+1] # length=window
    mdfit = LogisticRegression().fit(xold,y)
    predictResList.append(mdfit.predict(xnew)[0]) #使用最新数据，预测下一期，并保存在预测结果中
predictOutratio = float(np.sum((predictResList-Y[datewindow:])==0))/len(Y)*100
print('样本外预测准确率：%0.2f%%\n'% predictOutratio)
# 最后将预测结果保存到DataFrame中，在回测时会用到
predictRes = pd.DataFrame(predictResList,index=ttick,columns=['predictRes'])
predictRes['yyyymm'] = predictRes.index
predictRes.head()
```

```python
# step 7 : 有序多分类Logit回归
# 改进的地方：
# a. 针对此前未考虑到不显著变量进行优化，每次剔除不显著变量
# b. 将待预测指数收益差，从二分类细分为多分类，不仅考虑预测方向的问题，而且考虑预测涨幅的问题

# （1） 生成待预测的虚拟变量 Y
def myDummyVarGener_01(rawdata):
    '''
    大小盘风格差的虚拟变量，即 IF(（小盘月收益率-大盘月收益率）>0)
    将小盘股超出大盘股的收益，按照幅度分为5个档位：
    Dummy Y [-2：大幅跑输；-1：小幅跑输；0：持平；1：小幅跑赢；2：大幅跑赢]
    '''
    x = np.array(rawdata['mret_smb'])
    Q1 = np.percentile(x,20) # lower 20%
    Q2 = np.percentile(x,40)
    Q3 = np.percentile(x,60)
    Q4 = np.percentile(x,80) # top 20%
    Y = np.zeros_like(x)
    Y[np.where(x>Q4)]=2
    Y[np.where((x<=Q4)*(x>Q3))]=1 # sign '*' means 'and'
    Y[np.where((x<=Q2)*(x>Q1))]=-1
    Y[np.where(x<=Q1)]=-2
    return Y
# （2）因子标准化
# 本部分不做修改，沿用前值不变
# (3) 有序多分类Logit回归预测
# 新增：有效变量筛选
rawdata = myMacroDataload()
Y = myDummyVarGener_01(rawdata)
datewindow = 20 # 申万研究推荐采用20日作为滚动预测的时间窗口
predictResList0 = []
ttick = []
for t in range(0,len(Y)-datewindow): # implied t: start month to end-window month
    xraw = rawdata.iloc[t:t+datewindow+1] # iloc:按行取数据；t要+1，否则最后一个值取不到,length=window
    ttick.append(rawdata.index[t+datewindow]) # get current Date
    xold,xnew = myGenerFitdata(xraw)
    y = Y[t:t+datewindow+1] # length=window
    mdfit = LogisticRegression().fit(xold,y)
    # Uqer不支持statsmodel，无法剔除不显著的变量    
    predictResList0.append(mdfit.predict(xnew)[0]) #使用最新数据，预测下一期，并保存在预测结果中
predictOutratio0 = float(np.sum((predictResList0*Y[datewindow:])>0)+np.sum((np.array(predictResList0)==0)*(Y[datewindow:]==0)))/len(Y)*100
# predictOutratio0 = float(np.sum((predictResList0-Y[datewindow:])==0))/len(Y)*100
print('样本外预测准确率：%0.2f%%\n'% predictOutratio0)
# 最后将预测结果保存到DataFrame中，在回测时会用到
predictRes0 = pd.DataFrame(predictResList0,index=ttick,columns=['predictRes'])
predictRes0['yyyymm'] = predictRes0.index
predictRes0.head(30)
```
