# [WorldQuant的Alpha10]

新出了Signal模块，感觉很赞啊。写个worldQuant的一个Alpha和大家分享一下。

**Alpha#10:**
$$rank(((0 < ts\_min(delta(close, 1), 4)) ? delta(close, 1) : ((ts\_max(delta(close, 1), 4) < 0) ? delta(close, 1) : (-1 * delta(close, 1)))))$$
可以看出，这个公式的主要逻辑是：

股票最近5天是否一直上涨？---是，则取最近一天的收盘价与前一天的收盘价的差值。

----否，则继续判断股票最近5天是否一直下跌？---是，则取最近一天的收盘价与前一天的收盘价的差值。----否，则取最近一天的收盘价与前一天的收盘价的差值的负数。然后再将各股票对应的数值进行排序。

简单的分析可以发现，该公式大致思想是：买连续上涨的股票，昨日涨幅越大的股票越买；避免连续下跌的股票，昨日跌幅越小的股票越买；处于震荡的股票昨日跌幅越大买。
