### ADX指标

- ADX，全称Average Directional Index，是一种用于衡量资产行情是否存在趋势的技术指标
- ADX的计算方法十分复杂，涉及到+DM, -DM, TR, +DI, -DI等很多中间变量：
	- 首先计算+DM和-DM，即所谓上升动向与下降动向，公式为
	- $$+DM_t=\text{t日最高价 - (t-1)日最高价}\\-DM_t=\text{(t-1)日最低价 - t日最低价}$$
	- 其次我们来计算TR，即所谓的真实波幅：$TR_t=\max(|t日最高-t日最低|, |t日最高-(t-1)日收盘|, |t日最低-(t-1)日收盘|)$
	- 然后利用+DM,-DM,TR来计算DI：$\pm DM_t=\sum_{k=0}^{N-1}\pm DM_{t-k}/\sum_{k=0}^{N-1}TR_{t-k}\times100$，这里的N是参数，一般取14
	- 然后我们就可以计算ADX和ADXR了：
	- $$ADX_t=MA\left(\frac{(+DI_t)-(-DI_t)}{(+DI_t)+(-DI_t)}\times100,N\right)$$
