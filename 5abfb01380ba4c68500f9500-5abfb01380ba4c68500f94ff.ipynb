# 指数乖离率统计

乖离率，简称Y值，是移动平均原理派生的一项技术指标，其功能主要是通过测算股价在波动过程中与移动平均线出现偏离的程度，从而得出股价在剧烈波动时因偏离移动平均趋势而造成可能的回档或反弹，以及股价在正常波动范围内移动而形成继续原有趋势的可信度。

乖离度的测市原理是建立在：如果股价偏离移动平均线太远，不管股价在移动平均线之上或之下，都有可能趋向平均线的这一条原理上。而乖离率则表示股价偏离趋向指标占百分比值。

国内交易软件的乖离率指标一般都采用6日、12日、24日等短周期的均线，用以判断短周期的买卖时机，本文尝试从主要指数长周期均线（年线）入手，分析乖离率（以下均指相对年线的乖离率）提示中长期交易时点的可能性。

### 一、指数乖离率与业绩增速的假定分析
假设将指数看做一家公司，初始的基准点位为b，每年的业绩增速为g（g为常量），为简单化处理，这里暂时假定指数的估值水平恒定不变，则第n年指数的收盘点位可表示为：

$F(n) = b * (1+g)^n$		(1.1)

第n年的年线（250天均线）可由(F(n-1) + F(n)) / 2推导出：

$MA(n) = b * (1+g)^{n-1} * (2+g) / 2$	(1.2)

则第n年的乖离率即为(F(n) - MA(n)) / MA(n)，如下：

$B(n) = g / (2+g)$			(1.3)

函数作图如下：

```python
import numpy as np
import matplotlib.pyplot as plt

# 参数初始化
base = 1000               # 基准点位
g = 0.20                  # 净利润增长率（g>0）
n = 20                    # 时间周期长度（单位为年）
x = np.arange(0, n+1, 1)  # 时间周期序列

# 时间周期N收益函数
def close_n(base, g, x):
    close = base * (1 + g) ** x
    return close

# 收益序列
y_close = close_n(base, g, x)

# 均线序列
y_ma = (close_n(base, g, x) + close_n(base, g, x-1)) / 2

# 乖离率
y_bias = y_close / y_ma - 1

# 绘图
fig, ax1 = plt.subplots(figsize=(12,8))
ax1.plot(x, y_close, label="close_price")
ax1.plot(x, y_ma, label="ma_250")
ax1.legend(loc=2)
ax2 = ax1.twinx()
ax2.plot(x, y_bias, label="bias_ratio", color="gray")
ax2.legend(loc=1)
```

可以看出，在不考虑估值水平变化的情况下，指数的乖离率为常量，并由业绩增速g决定。假设业绩增速为20%，则指数乖离率为9.09%，**即指数的收盘点位稳定在年线上方9.09%的位置（与通常认为指数向年线回归的习惯性认知稍有区别）。** 同时在已知乖离率时，也可由公式1.3反向推导出业绩增速。

统计结果显示，向下偏离极值方面，除创业板指未经过08年的大跌，其他指数的最小值都超过50%，而向上偏离极值方面，则从90%到140%不等，反映了A股急涨急跌的高波动性。参考成熟股市的数据，如美股的道琼斯指数，乖离率指标在-30%到30%之间波动，反映了成熟市场的低波动性。

**乖离率的均值方面，介于4.18%（上证综指）~11.49%（中证1000），即从长期来看，指数趋向于年线上方的特定位置回归，与公式1.3的推导结果一致。** 对比指数的实际业绩增速，上证综指、沪深300、中证500的推导业绩增速与实际情况较为符合，不一致的指数中，如创业板的推导业绩增速偏低（可能与统计周期较短有关），中证1000推导业绩增速偏高（可能与小市值股票的高溢价有关），因为公式暂未考虑估值水平的变动因素，尚需要进一步的研究。

```python
import numpy as np
import pandas as pd
import DataAPI
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats, integrate

# 计算指数乖离率函数
def get_index_bias(index, start_date, end_date):
    # 获取价格序列
    df = DataAPI.MktIdxdGet(indexID=index, beginDate=start_date, endDate=end_date, field=['tradeDate', 'closeIndex'])
    df['tradeDate'] = pd.to_datetime(df['tradeDate'])
    df = df.set_index('tradeDate')

    # 获取均线数据，计算乖离率
    window_short = 20
    window_long = 250
    df['short_window'] = np.round(pd.rolling_mean(df['closeIndex'], window=window_short), 2)
    df['long_window'] = np.round(pd.rolling_mean(df['closeIndex'], window=window_long), 2)
    df['bias_ratio'] = np.round((df.closeIndex-df.long_window)/df.long_window, 4)

    # 去除乖离率nan值
    df['bias_nan'] = np.isnan(df.bias_ratio)
    df = df[df.bias_nan == False]
    
    return df

# 指数列表
index_list = ('000001.ZICN', #上证指数
              '399001.ZICN', #深证成数
              '399005.ZICN', #中小板指
              '399006.ZICN', #创业板指
              '000016.ZICN', #上证50
              '000300.ZICN', #沪深300
              '000905.ZICN', #中证500
              '000852.ZICN', #中证1000
             )
start_list = ('1992-01-01', #上证指数
              '1992-01-01', #深证成数
              '2006-02-01', #中小板指
              '2010-06-01', #创业板指
              '2004-01-01', #上证50
              '2002-01-01', #沪深300
              '2005-01-01', #中证500
              '2005-01-01', #中证1000
             )

# 初始化指数DataFrame，包括指数名称和样本数据起止日期
df_index = pd.DataFrame(index=index_list, columns=('name','start_date','end_date','bias_max','bias_min','bias_mean','bias_std','g'))
df_index['name'] = [DataAPI.IdxGet(secID=index).secShortName[0] for index in index_list]
df_index['start_date'] = start_list
df_index['end_date'] = '2017-12-31'

# 添加乖离率均值和方差
for index in index_list:
    df_bias = get_index_bias(index, df_index.start_date[index], df_index.end_date[index])
    df_index.bias_max[index] = np.round(df_bias.bias_ratio.max(), 4)
    df_index.bias_min[index] = np.round(df_bias.bias_ratio.min(), 4)
    df_index.bias_mean[index] = np.round(df_bias.bias_ratio.mean(), 4)
    df_index.bias_std[index] = np.round(df_bias.bias_ratio.std(), 4)
    df_index.g[index] = 2 * np.round(df_index.bias_mean[index] / (1 - df_index.bias_mean[index]), 4)

df_index
```

```python
# 初始参数
index = '000001.ZICN'
security = get_index_bias(index, df_index.start_date[index], df_index.end_date[index])
security.head()

# 价格序列可视化
fig, ax1 = plt.subplots(figsize=(20,8))
ax1.plot(security.index, security.closeIndex, security.index, security.long_window)

ax2 = ax1.twinx()
ax2.plot(security.index, security.bias_ratio, color="gray")
ax2.legend(loc=1)

plt.show()
```

概率分布方面，参考上证综指的历史分布，红色区域为实际分布，蓝色曲线为基于乖离率均值和方差的正态分布，即实际分布类似于正态分布，但不完全一致，在均值附近分布较为集中，靠近极值区域，分布概率较低，符合“如果股价偏离移动平均线太远，不管股价在移动平均线之上或之下，都有可能趋向平均线”这一原理。

```python
# 指数乖离率分布图
fig = plt.figure(figsize = (16, 8))

# 直方图
# sns.distplot(security.bias_ratio, kde=False, rug=True)

# 核密度估计
sns.kdeplot(security.bias_ratio, label = 'bias_ratio', color="#FF0000", alpha=.6, shade=True)

# 拟合参数分布
# sns.distplot(security.bias_ratio, kde=False, fit=stats.gamma)

# 正态分布
bias_min = security.bias_ratio.min()
bias_max = security.bias_ratio.max()
mu = security.bias_ratio.mean()
sigma = security.bias_ratio.std()
x = np.arange(bias_min, bias_max, 0.01)
y = stats.norm.pdf(x, mu, sigma)
plt.plot(x, y)
```

基于以上的分析，同时考虑A股的波动特性，结合指数的历史估值水平，使用乖离率的极值可以提示中长线的转折点。
