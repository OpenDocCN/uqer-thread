```python
from CAL.PyCAL import *    # CAL.PyCAL中包含font
import numpy as np
import pandas as pd
import time
import lib.checkuniverse
import matplotlib
import matplotlib.pyplot as plt
from matplotlib import font_manager as fm

from pandas import Series, DataFrame
from datetime import datetime, timedelta

#===============================================
#获取年报基本面数据
#===============================================
def get_fund_data(tickerID=u"",reportType=["A"],lastYears=5,endDate=u"",beginDate=u""):
    
    asset_df = DataAPI.FdmtBSInduAllLatestGet(ticker=tickerID,reportType=reportType,
                                              endDate=endDate,beginDate=beginDate,pandas="1").head(lastYears)
    is_df    = DataAPI.FdmtISInduAllLatestGet(ticker=tickerID,reportType=reportType,
                                               endDate=endDate,beginDate=beginDate,pandas="1").head(lastYears)
    cf_df    = DataAPI.FdmtCFInduAllLatestGet(ticker=tickerID,reportType=reportType,
                                               endDate=endDate,beginDate=beginDate,pandas="1").head(lastYears)
    
    asset_df['Year'] = asset_df.endDate.apply(lambda x: x[0:4])
    asset_df.Year = asset_df.Year + asset_df.reportType 
    
    return asset_df    

#===============================================
#获取滚动的年报基本面数据
#===============================================

def get_fund_data_2(tickerID=u"",lastYears=8,endDate=u"",beginDate=u""):
    
    #获取最后一年的年报数据
    df_year = DataAPI.FdmtBSInduAllLatestGet(ticker=tickerID,reportType=["A"],
                                              endDate=endDate,beginDate=beginDate,pandas="1")
    if df_year.index.size < 1:
        df = DataAPI.FdmtBSInduAllLatestGet(ticker=tickerID,reportType=['Q1','S1','Q3','A'],
                                              endDate=endDate,beginDate=beginDate,pandas="1").head(lastYears)
        print  u'无年报'
        df['Year'] = df.endDate.apply(lambda x: x[0:4])
        df.Year = df.Year + df.reportType 
        return df
    
    #获取最后一年年报之后的季报数据
    df_q = DataAPI.FdmtBSInduAllLatestGet(ticker=tickerID,reportType=['Q1','S1','Q3','A'],
                                            endDate=endDate,beginDate=beginDate,pandas="1")
    lastYearDate = df_year.endDate[0]
    df_q = df_q[df_q.endDate > lastYearDate]
    if df_q.index.size < 1:
        df =  df_year.head(lastYears)
        print  u'无季报'
    else:
        #合并季报年报数据
        df = df_q.copy();
        df = df.append(df_year,ignore_index=True).head(lastYears)
    
    df['Year'] = df.endDate.apply(lambda x: x[0:4])
    df.Year = df.Year + df.reportType 
    
    return df    

#===============================================
#自动居中标注
#===============================================
def autolabel(ax, rects,texts):
    
    pos = 0
    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x()+rect.get_width()/2., rect.get_y()+0.5*height, texts[pos],
                ha='center', va='center')
        pos = pos + 1
        
    return
   


#===============================================    
#资产负债表分析 --- 流动资产结构图
#===============================================
def parse_asset_LDZC(asset_df,ax2,limit= 0.01):
    
    #获取第一行
    row= asset_df.iloc(0)[0]
    
    #创建DataFrame
    name  = ['cashCEquiv','settProv','loanToOthBankFi','tradingFA','NotesReceiv',
             'AR','prepayment','premiumReceiv','reinsurReceiv',
             'reinsurReserReceiv','intReceiv','divReceiv','othReceiv',
             'purResaleFa','inventories','NCAWithin1Y','othCA']
        
  
    value  = [row.cashCEquiv,row.settProv,row.loanToOthBankFi,row.tradingFA,row.NotesReceiv,
             row.AR,row.prepayment,row.premiumReceiv,row.reinsurReceiv,
             row.reinsurReserReceiv,row.intReceiv,row.divReceiv,row.othReceiv,
             row.purResaleFa,row.inventories,row.NCAWithin1Y,row.othCA]
    

    label = [u'货币资金',u'结算备付金',u'拆出资金',u'交易性金融资产',u'应收票据',
             u'应收账款',u'预付款项',u'应收保费',u'应收分保账款',
             u'应收分保合同准备金',u'应收利息',u'应收股利',u'其他应收款',
             u'买入返售金融资产',u'存货',u'一年内到期的非流动资产',u'其他流动资产']
    
    result = DataFrame(columns=['Value','Percent','Label'],
                       index=name)
    
    #填充数据
    for pos in range(0,len(name)):
        result.loc[ name[pos] ] = [ value[pos], value[pos]/row.TCA, label[pos] ]
    result = result[result.Percent >= limit]
    result.loc['Other'] = [row.TCA-result.Value.sum(), 1 - result.Value.sum()/row.TCA,u'其它合计' ]
    result = result.sort(columns=['Percent'])
    #print result
    #print result.Percent.sum(),result.Value.sum(),row.TCA

    #出图
    ax2.grid(True)
    ax2.set_title( u'流动资产结构',fontproperties = font, fontsize = 16)
    ax2.set_xlabel(row.Year, fontsize = 16)
    patches, texts, autotexts =ax2.pie(result.Percent.tolist(), labels= result.Label.tolist(),
                autopct='%1.1f%%',pctdistance=0.8, shadow=True,colors  = ["pink","coral","yellow","orange"])
         
    plt.setp(autotexts, fontproperties=font,fontsize=16)
    plt.setp(texts, fontproperties=font,fontsize=16)        
    return

#===============================================    
#资产负债表分析 --- 非流动资产结构图
#===============================================
def parse_asset_FLDZC(asset_df,ax3,limit= 0.01):
    
    #获取第一行
    row= asset_df.iloc(0)[0]
    
    #创建DataFrame
    name  = ['htmInvest','LTEquityInvest','investRealEstate','fixedAssets','CIP',
             'constMaterials','fixedAssetsDisp','producBiolAssets','intanAssets',
             'goodwill','LTAmorExp','deferTaxAssets','othNCA',
             'LTReceive','oilAndGasAssets','RD']
  
    value  = [row.htmInvest,row.LTEquityInvest,row.investRealEstate,row.fixedAssets,row.CIP,
             row.constMaterials,row.fixedAssetsDisp,row.producBiolAssets,row.intanAssets,
             row.goodwill,row.LTAmorExp,row.deferTaxAssets,row.othNCA,
             row.LTReceive,row.oilAndGasAssets,row.RD]
    

    label = [u'持有至到期投资',u'长期股权投资',u'投资性房地产',u'固定资产',u'在建工程',
             u'工程物资',u'固定资产清理',u'生产性生物资产',u'无形资产',
             u'商誉',u'长期待摊费用',u'递延所得税资产',u'其他非流动资产',
             u'长期应收款',u'油气资产',u'开发支出']
    
    result = DataFrame(columns=['Value','Percent','Label'],
                       index=name)
    
    #填充数据
    for pos in range(0,len(name)):
        result.loc[ name[pos] ] = [ value[pos], value[pos]/row.TNCA, label[pos] ]
    result = result[result.Percent >= limit]
    result.loc['Other'] = [row.TNCA-result.Value.sum(), 1 - result.Value.sum()/row.TNCA,u'其它合计' ]
    result = result.sort(columns=['Percent'])
    #print result
    #print result.Percent.sum(),result.Value.sum(),row.TNCA

    #出图
    ax3.grid(True)
    ax3.set_title( u'非流动资产结构',fontproperties = font, fontsize = 16)
    ax3.set_xlabel(row.Year, fontsize = 16)
    patches, texts, autotexts =ax3.pie(result.Percent.tolist(), labels= result.Label.tolist(),
                autopct='%1.1f%%',pctdistance=0.8, shadow=True,colors  = ["pink","coral","yellow","orange"])
         
    plt.setp(autotexts, fontproperties=font,fontsize=16)
    plt.setp(texts, fontproperties=font,fontsize=16)        
    return

#===============================================    
#资产负债表分析 --- 流动负债结构图
#===============================================
def parse_asset_LDFZ(asset_df,ax4,limit= 0.01):
    
    #获取第一行
    row= asset_df.iloc(0)[0]
    
    #创建DataFrame
    name  = ['STBorr','CBBorr','depos','loanFrOthBankFi','tradingFL',
             'NotesPayable','AP','advanceReceipts','soldForRepurFa',
             'commisPayable','payrollPayable','taxesPayable','intPayable',
             'divPayable','othPayable','reinsurPayable','insurReser',
             'fundsSecTradAgen','fundsSecUndwAgen','NCLWithin1Y','othCL']
  
    value  = [row.STBorr,row.CBBorr,row.depos,row.loanFrOthBankFi,row.tradingFL,
             row.NotesPayable,row.AP,row.advanceReceipts,row.soldForRepurFa,
             row.commisPayable,row.payrollPayable,row.taxesPayable,row.intPayable,
             row.divPayable,row.othPayable,row.reinsurPayable,row.insurReser,
             row.fundsSecTradAgen,row.fundsSecUndwAgen,row.NCLWithin1Y,row.othCL]
    

    label = [u'短期借款',u'向中央银行借款',u'吸收存款及同业存放',u'拆入资金',u'交易性金融负债',
             u'应付票据',u'应付账款',u'预收款项',u'卖出回购金融资产款',
             u'应付手续费及佣金',u'应付职工薪酬',u'应交税费',u'应付利息',
             u'应付股利',u'其他应付款',u'应付分保账款',u'保险合同准备金',
             u'代理买卖证券款',u'代理承销证券款',u'一年内到期的非流动负债',u'其他流动负债']
    
    result = DataFrame(columns=['Value','Percent','Label'],
                       index=name)
    
    #填充数据
    for pos in range(0,len(name)):
        result.loc[ name[pos] ] = [ value[pos], value[pos]/row.TCL, label[pos] ]
    result = result[result.Percent >= limit]
    result.loc['Other'] = [row.TCL-result.Value.sum(), 1 - result.Value.sum()/row.TCL,u'其它合计' ]
    result = result.sort(columns=['Percent'])
    #print result
    #print result.Percent.sum(),result.Value.sum(),row.TCL

    #出图
    ax4.grid(True)
    ax4.set_title( u'流动负债结构',fontproperties = font, fontsize = 16)
    ax4.set_xlabel(row.Year, fontsize = 16)
    patches, texts, autotexts =ax4.pie(result.Percent.tolist(), labels= result.Label.tolist(),
                autopct='%1.1f%%',pctdistance=0.8, shadow=True,colors  = ["pink","coral","yellow","orange"])
         
    plt.setp(autotexts, fontproperties=font,fontsize=16)
    plt.setp(texts, fontproperties=font,fontsize=16)        
    return

#===============================================    
#资产负债表分析 --- 非流动负债结构图
#===============================================
def parse_asset_FLDFZ(asset_df,ax5,limit= 0.01):
    
    #获取第一行
    row= asset_df.iloc(0)[0]
    
    #创建DataFrame
    name  = ['LTBorr','bondPayable','LTPayable','specificPayables','estimatedLiab',
             'deferTaxLiab','othNCL']
  
    value  = [row.LTBorr,row.bondPayable,row.LTPayable,row.specificPayables,row.estimatedLiab,
             row.deferTaxLiab,row.othNCL]
   
    label = [u'长期借款',u'应付债券',u'长期应付款',u'专项应付款',u'预计负债',
             u'递延所得税负债',u'其他非流动负债']
    
    result = DataFrame(columns=['Value','Percent','Label'],
                       index=name)
    
    #填充数据
    for pos in range(0,len(name)):
        result.loc[ name[pos] ] = [ value[pos], value[pos]/row.TNCL, label[pos] ]
    result = result[result.Percent >= limit]
    result.loc['Other'] = [row.TNCL-result.Value.sum(), 1 - result.Value.sum()/row.TNCL,u'其它合计' ]
    result = result.sort(columns=['Percent'])
    #print result
    #print result.Percent.sum(),result.Value.sum(),row.TNCL

    #出图
    ax5.grid(True)
    ax5.set_title( u'非流动负债结构',fontproperties = font, fontsize = 16)
    ax5.set_xlabel(row.Year, fontsize = 16)
    patches, texts, autotexts =ax5.pie(result.Percent.tolist(), labels= result.Label.tolist(),
                autopct='%1.1f%%',pctdistance=0.8, shadow=True,colors  = ["pink","coral","yellow","orange"])
         
    plt.setp(autotexts, fontproperties=font,fontsize=16)
    plt.setp(texts, fontproperties=font,fontsize=16)        
    return


#===============================================    
#资产负债表分析 --- 所有者权益结构图
#===============================================
def parse_asset_SYZQY(asset_df,ax6,limit= 0.01):
    
    #获取第一行
    row= asset_df.iloc(0)[0]
    
    #创建DataFrame
    name  = ['paidInCapital','capitalReser','treasuryShare','specialReser','surplusReser',
             'ordinRiskReser','retainedEarnings','forexDiffer']
  
    value  = [row.paidInCapital,row.capitalReser,-row.treasuryShare,row.specialReser,row.surplusReser,
             row.ordinRiskReser,row.retainedEarnings,row.forexDiffer]
    

    label = [u'实收资本',u'资本公积',u'减:库存股',u'专项储备',u'盈余公积',
             u'一般风险准备',u'未分配利润',u'外币报表折算差额']    

    result = DataFrame(columns=['Value','Percent','Label'],
                       index=name)
    
    #填充数据
    for pos in range(0,len(name)):
        result.loc[ name[pos] ] = [ value[pos], value[pos]/row.TEquityAttrP, label[pos] ]
    result = result[result.Percent >= limit]
    result.loc['Other'] = [row.TEquityAttrP-result.Value.sum(), 1 - result.Value.sum()/row.TEquityAttrP,u'其它合计' ]
    result = result.sort(columns=['Percent'])
    #print result
    #print result.Percent.sum(),result.Value.sum(),row.TEquityAttrP

    #出图
    ax6.grid(True)
    ax6.set_title( u'归属母公司所有者权益结构',fontproperties = font, fontsize = 16)
    ax6.set_xlabel(row.Year, fontsize = 16)
    patches, texts, autotexts =ax6.pie(result.Percent.tolist(), labels= result.Label.tolist(),
                autopct='%1.1f%%',pctdistance=0.8, shadow=True,colors  = ["pink","coral","yellow","orange"])
         
    plt.setp(autotexts, fontproperties=font,fontsize=16)
    plt.setp(texts, fontproperties=font,fontsize=16)        
    return

#===============================================    
#资产负债表分析 --- 总资产、净资产变动图
#===============================================
def parse_asset_ZCBD(asset_df,ax7):
    
    #出图
    ax7.grid(True)
    ax7.set_title( u'资产变动',fontproperties = font, fontsize = 16)
    ax7.set_ylabel(u'百万', fontproperties = font, fontsize = 16)
    
    x = range(asset_df.Year.size,0,-1)
    years = asset_df.Year.tolist()
    
    l1 = ax7.plot(x, asset_df.TAssets/1e6, marker='o')
    l2 = ax7.plot(x, asset_df.TEquityAttrP/1e6, marker='o')
    ax7.set_xticks( x )
    ax7.set_xticklabels(years)    

    #图例    
    leg_text = [u'总资产',u'净资产']
    leg = ax7.legend(leg_text,2,prop=font,fontsize=16)
    leg.get_frame().set_alpha(0.5) 

    #标注
    x = l1[0].get_xdata()
    y = l1[0].get_ydata()
    for i in range(0,len(x)-1):
        ax7.text(x[i], y[i]*1.01, str(int(y[i])) + '( '+ str(round((y[i]/y[i+1]-1)*100,1)) +'% )',
                 ha='center', va='bottom',color='b')
        
    x = l2[0].get_xdata()
    y = l2[0].get_ydata()
    for i in range(0,len(x)-1):
        ax7.text(x[i], y[i]*1.01, str(int(y[i])) + '( '+ str(round((y[i]/y[i+1]-1)*100,1)) +'% )',
                 ha='center', va='bottom',color='g')        
        
    return

#===============================================    
#资产负债表分析 --- 资产负债表结构图
#===============================================
def parse_asset_ZCGGB2(df,ax1,showlengend=True):
    
    asset_df = df.fillna(0)
    
    #------------- 资产负债结构图 ----------
    bw = 0.4
    bo = -0.4
    yuan_unit = 1e6
    c =['orange','y','orange','y','r']
    pos = asset_df.index[0]
    
    ax1.grid(True)
    ax1.set_title( u'资产负债表结构',fontproperties = font, fontsize = 16)
    ax1.set_ylabel(u"百万",fontproperties = font, fontsize = 16)
    ax1.set_xlabel(asset_df.secShortName[pos].decode("utf-8"),fontproperties = font, fontsize = 16)
    
    #流动资产
    rc_TCA = ax1.bar(left = asset_df.index.size - asset_df.index + bo  ,height =asset_df.TCA / yuan_unit ,
                     bottom = asset_df.TNCA / yuan_unit , width=bw,color= c[0],alpha = .5)
    text =  asset_df.TCA/(asset_df.TNCA+asset_df.TCA)*100+0.5
    text =  text.apply(lambda x: str(int(x))+"%")
    autolabel(ax1, rc_TCA, text)
    
    #非流动资产
    rc_TNCA = ax1.bar(asset_df.index.size - asset_df.index + bo  ,asset_df.TNCA / yuan_unit,width=bw,color= c[1],alpha = .5)
    text =  asset_df.TNCA/(asset_df.TNCA+asset_df.TCA)*100+0.5
    text =  text.apply(lambda x: str(int(x))+"%")
    autolabel(ax1, rc_TNCA, text)
 
    #处理NAN


    
    #流动负债
    rc_TCL = ax1.bar(left=asset_df.index.size - asset_df.index + bo + bw ,height = asset_df.TCL/ yuan_unit,
                      bottom =  (asset_df.TNCL + asset_df.TShEquity)/yuan_unit,width=bw, color= c[2], alpha = .3 )
    text =  asset_df.TCL/(asset_df.TNCL + asset_df.TShEquity+ asset_df.TCL)*100+0.5
    text =  text.apply(lambda x: str(int(x))+"%")
    autolabel(ax1, rc_TCL, text)
    
    #非流动负债
    rc_TNCL = ax1.bar(left=asset_df.index.size - asset_df.index + bo + bw ,height = asset_df.TNCL / yuan_unit,
                      bottom =asset_df.TShEquity/yuan_unit, width=bw,color= c[3],alpha = .3)
    text =  asset_df.TNCL/(asset_df.TNCL + asset_df.TShEquity+ asset_df.TCL)*100+0.5
    text =  text.apply(lambda x: str(int(x))+"%")
    autolabel(ax1, rc_TNCL, text)   
    
    #所有者权益
    rc_TE = ax1.bar(asset_df.index.size - asset_df.index + bo + bw ,asset_df.TShEquity/ yuan_unit, width=bw,color= c[4],alpha = .3)
    text =  asset_df.TShEquity/(asset_df.TNCL + asset_df.TCL + asset_df.TShEquity)*100+0.5
    text =  text.apply(lambda x: str(int(x))+"%")
    autolabel(ax1, rc_TE, text)
    
    #图例
    years = asset_df.Year.tolist()
    years.reverse()
    ax1.set_xticks( (asset_df.index + 1).tolist() )
    ax1.set_xticklabels(years)    
    
    if showlengend :
        leg_text = [u'流动资产',u'非流动资产',u'流动负债',u'非流动负债',u'所有者权益']
        leg = ax1.legend([rc_TCA[0],rc_TNCA[0],rc_TCL[0],rc_TNCL[0],rc_TE[0]],leg_text,'best',prop=font,fontsize=16)
        leg.get_frame().set_alpha(0.5) 
    
    #总资产标注
    text =  (asset_df.TNCA+asset_df.TCA)/yuan_unit
    text =  text.apply(lambda x:round(float(x), 2))    
    pos = 0
    for rect in rc_TCA:
        height = rect.get_height()
        ax1.text(rect.get_x()+rect.get_width(), rect.get_y()+1.01*height, text[pos],ha='center', va='bottom')
        pos = pos + 1    
    
    return


```
