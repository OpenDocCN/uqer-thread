```python
import quartz
from quartz.api import *
from quartz import backtest
import pandas as pd
import numpy as np
from datetime import datetime
from CAL.PyCAL import Calendar, Date, Period, BizDayConvention
from matplotlib import pylab


start = '2013-05-01'                       # 回测起始时间
end = u'2016-09-06'                         # 回测结束时间
benchmark = 'HS300'                        # 策略参考标准
universe = ['510300.XSHG','510500.XSHG','511880.XSHG']  # 证券池，支持股票和基金
capital_base = 10000                      # 起始资金
freq = 'd'                                 # 策略类型，'d'表示日间策略使用日线回测，'m'表示日内策略使用分钟线回测
refresh_rate = 1                           # 调仓频率，表示执行handle_data的时间间隔，若freq = 'd'时间间隔的单位为交易日，若freq = 'm'时间间隔为分钟
commission = Commission(0.0, 0.0)

def initialize(account):                   # 初始化虚拟账户状态
    pass

def handle_data(account):                  # 每个交易日的买入卖出指令
    
    cur_date = Date.fromDateTime(account.current_date)
    cal = Calendar("China.SSE")
    pre_day = cal.advanceDate(cur_date, '-1B', BizDayConvention.Preceding).strftime('%Y-%m-%d')
    pre_20day = cal.advanceDate(cur_date, '-21B', BizDayConvention.Preceding).strftime("%Y-%m-%d")
    
    df = DataAPI.MktIdxdGet(tradeDate=u"",indexID=u"",ticker=u"000300,000905",beginDate=pre_20day,endDate=pre_day,field=u"tradeDate,ticker,closeIndex",pandas="1")
    df = df.pivot(index="tradeDate", columns="ticker", values="closeIndex")
    
    s300 = df["000300"]
    s500 = df["000905"]
    
    pct_chg300 = (s300[pre_day] - s300[pre_20day]) / (s300[pre_20day])
    pct_chg500 = (s500[pre_day] - s500[pre_20day]) / (s500[pre_20day])
    
    if pct_chg300 < 0 and pct_chg500 < 0:
        order_pct_to("510300.XSHG", 0)
        order_pct_to("510500.XSHG", 0)
        order_pct_to("511880.XSHG", 0.95)
    elif pct_chg300 > pct_chg500:
        order_pct_to("510300.XSHG", 0.95)
        order_pct_to("510500.XSHG", 0)
        order_pct_to("511880.XSHG", 0)  
    elif pct_chg500 > pct_chg300:
        order_pct_to("510300.XSHG", 0)
        order_pct_to("510500.XSHG", 0.95)
        order_pct_to("511880.XSHG", 0)  

bt, acct = backtest(start, end, benchmark, universe, capital_base, initialize, handle_data, refresh_rate=refresh_rate, freq=freq, commission=commission)

perf = quartz.perf_parse(bt, acct)
perf['cumulative_returns'].plot()
perf['benchmark_cumulative_returns'].plot()
# pylab.legend(['current_strategy', 'HS300'])
```
