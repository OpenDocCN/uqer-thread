# 基础动量策略

**<center>动量因子研究</center>**
=

**本文复刻了石川老师高质量动量选股策略，在Uqer回测框架下，动量因子长期无法显著跑赢大盘，持续低于指数表现，也说明了A股市场传统单因子策略无法获得持续超额收益**
2013年，Asness在JF杂志上发表了著名的《Value and Momentum everywhere》，证明了动量因子和价值因子在各类资产大类之间都普遍存在，且动量因子和价值因子存在明显负相关关系。因此，在中国市场研究动量因子非常有必要。
Asness的文章中，用过去12个月（不包括最近一个月）的收益率作为衡量股票因子的动量，显然，这在A股市场显然不适用，A股市场的涨跌幅限制和散户占多数的市场结构使得A股市场并不像美国市场那样更贴近“有效市场”。
因此，依据石川老师的分享，我们在这里用某只股票**过去六十个交易日的收益率减去其过去六十日的方差作为动量的代理指标**。即：
$$Momentum=r_{60}-3000*\sigma_{60}^2$$
其中，3000是厌恶系数，用以平滑收益率和方差之间量级上的差异。

**显然，单从这一简单的动量代理指标，无法完成进一步的筛选**
比如，对于同一时点，两只“动量”相似的股票，我们无法判断其“动量”来源的路径，即，我们无法通过这一单一指标判断其高动量是来自近日较大上行波动，还是来自低波动的持续股价上涨。
**很明显，这两种动量累积路径是不一样，也反应了不一样的市场情绪，不同的股票特征**
因此，我们希望进一步完成筛选，去选出那些真正的“高质量型的动量”，即捕获那些来自低波动的持续股价上行的高动量。
我们引入了两个代理指标，**MAX和ID**
**MAX**：我们取单只证券过去20日的最大日收益作为MAX指标，显然，在上述逻辑下，MAX因子值越小越好
**ID**：我们用过去60个交易日换手率的标准差作为ID指标，来反应期间股价波动情况，显然，ID指标也是越小越好

关于MAX和ID指标的理论依据是行为金融学中著名的“有限认知”，以及人们通常高估极端事件发生的概率，有关这方面的理论，还请自行参考教科书

**导入常用的Python第三方包**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from collections import OrderedDict
import statsmodels.api as sm
import seaborn as sns
from CAL.PyCAL import *
sns.set_style('whitegrid')
cal = Calendar('China.SSE')
```

**第一步：调取回测期间内的月末交易日，每日调仓**
=
按照上述逻辑，我们先要提取当天的日期，20个交易日前，以及60个交易日前的时间日期序列

```python
def get_order_date(start,end):
    field = ['calendarDate','isMonthEnd']
    temp_ls = []
    date = DataAPI.TradeCalGet(exchangeCD='XSHG',beginDate=start,endDate=end,isOpen='1',field=field,pandas='1') #提取日期序列
    for index,row in date.iterrows():
        if row['isMonthEnd'] == 1:
            temp_ls.append(row['calendarDate'])
        else:
            pass
    ls = list(set(temp_ls))   #提取的序列用重复，去重
    ls.sort(key=temp_ls.index)
    return ls
start = '2010-01-01'
end = '2020-01-01'
dates = get_order_date(start,end)
dates_60bef = [cal.advanceDate(date,'-60D',BizDayConvention.Following).strftime('%Y-%m-%d') for date in dates]  #前60交易日
dates_20bef = [cal.advanceDate(date,'-20D',BizDayConvention.Following).strftime('%Y-%m-%d') for date in dates]  #前20交易日
print(dates[:7])
print(dates_60bef[:7])
print(dates_20bef[:7])
```

**第二步，获取指数成分构成**
=
我们这里的证券池选用沪深300证券池

```python
def get_index_consID(date):
	date = date
	field = ['ticker','secShortName','consID','consShortName']
	index_data = DataAPI.IdxConsGet(ticker='000300',\
		intoDate=date,field=field,pandas='1')
	consID = list(index_data['consID'])
	return consID
```

**第三步，构建动量因子**
=
我们在函数中，一次性处理Momentum,ID,MAX三个因子的数据，返回类型为高级索引的DF格式，方便后续操作

```python
def get_momentum_factor(date_60bef,date_20bef,date):
    field = ['secID','tradeDate','chgPct','turnoverRate']
    dict_ = OrderedDict() #创建顺序构成的字典
    for date_60bef,date_20bef,date in list(zip(date_60bef,date_20bef,date)):
        code = []
        momentum = []
        ID = []
        Max = []
        t_stock = get_index_consID(date)
        s_stock = DataAPI.SecHaltGet(beginDate=date_60bef,endDate=date,listStatusCD='S',assetClass='E',field='secID',pandas='1')
        st_stockdf = DataAPI.SecSTGet(beginDate=date_60bef,endDate=date,field='secID',pandas="1")
        st_list = list(set(list(st_stockdf['secID'])))
        s_stock = list(s_stock['secID'])
        stock1 = [i for i in t_stock if i not in s_stock]
        stock = [i for i in stock1 if i not in st_list]
        temp_df = pd.DataFrame()
        for i in stock:
            sor1_df = DataAPI.MktEqudGet(secID=i,beginDate=date_60bef,endDate=date,field=field,pandas='1',isOpen=1)
            sor2_df = DataAPI.MktEqudGet(secID=i,beginDate=date_20bef,endDate=date,field=field,pandas='1',isOpen=1)
            if (sor1_df.empty|sor2_df.empty):
                continue
            df1 = sor1_df.copy()
            df2 = sor2_df.copy()
            df1 = df1.set_index('tradeDate')
            df2 = df2.set_index('tradeDate')
            temp_std = np.std(df1['chgPct'])
            temp_ri = np.prod(df1['chgPct'])
            Max.append(df2['chgPct'].max())
            ID.append(df1['turnoverRate'].std())
            momentum.append(temp_ri-3000*temp_std**2)
            code.append(df1.iloc[0,0])
        temp_df['codes'] = code
        temp_df['momentum'] = momentum
        temp_df['ID'] = ID
        temp_df['Max'] = Max
        temp_df = temp_df.set_index('codes')
        dict_[date] = temp_df
        print(date,'success')  #由于函数运行时间较长，每处理完一天的数据，输出一次
    datas = pd.concat(dict_.values(),keys=dict_.keys())
    datas.index.names = ['Date','Codes']
    return datas
```

```python
df = get_momentum_factor(dates_60bef,dates_20bef,dates)
df.to_csv('10y_3factor_hs300.csv') #此处过程较长，转入CSV存取
```

读取CSV文件，展示处理好的数据

```python
data = pd.read_csv('10y_3factor_hs300.csv',index_col = [0,1])
data.head().append(data.tail())
```

**计算综合得分**
=
我们按照因子打分法，没有经过标准化和去极值处理，主要因为这三个因子指标本身就限制在一定范围内，去极值等意义不大

```python
def get_score(data):
    rank_data1 = data['momentum'].rank(ascending=True) #动量因子升序排列，rank名次越大越好
    rank_data2 = data['ID'].rank(ascending = False)#ID因子降序排列，rank名次越大越好
    rank_data3 = data['Max'].rank(ascending = False)#MAc因子降序排列，rank名次越大越好
    rank_data = pd.concat([rank_data1,rank_data2,rank_data3],axis=1)
    data['Score'] = rank_data.mean(axis=1) #score分数越大越好
    return data
```

```python
score_data = data.groupby(level=0).apply(get_score) #按时间进行打分处理
score_data.head().append(score_data.tail())
```

**对动量因子进行分组处理，分五组**
=

```python
def get_group(data,factor='momentum',num=5):   #分组函数，组号越大越好
    rank = data[factor].rank(ascending=True)
    label = ['G'+str(i) for i in range(1,num+1)]
    temp_data = pd.cut(rank,bins=num,labels=label)
    temp_data.name = 'Group'
    new_data = data.join(temp_data)
    return new_data
```

```python
new_data = score_data.groupby(level=0).apply(get_group,factor='Score')
new_data.head(8).append(new_data.tail(8))
```

```python
group_data = new_data[new_data['Group'] == 'G5']   #我们仅测试第五组，理论上应该是表现最好的一组
print(group_data.head(10))
print(len(list(group_data.loc['2018-01-31'].index)))
tp_ls = list(group_data.index.get_level_values(0))
ls = list(set(tp_ls))
ls.sort(key=tp_ls.index)
```
