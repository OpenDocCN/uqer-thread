# 最优仓位-蒙特·卡罗方法

蒙特·卡罗方法（Monte Carlo method），也称统计模拟方法，是二十世纪四十年代中期由于科学技术的发展和电子计算机的发明，而被提出的一种以概率统计理论为指导的一类非常重要的数值计算方法。是指使用随机数（或更常见的伪随机数）来解决很多计算问题的方法。与它对应的是确定性算法。

下面是用蒙特·卡罗方法求解投机的最优仓位。假设已知投机的胜率p， 获胜的净回报率为a， 失败的净损失为b。 通俗的案例，有一个赌局是投硬币， 投出正面，会获得双倍的奖励。投出是反面，将会损失所有硬币。很明显这是一个期望为正的赌局，但是如果把所有本金都压入，将会损失所有本金的。如果下注太少， 那么本金的增长会过慢。那么多少的下注是最优的呢？ 这个最优下注的复合增长率是多少呢？

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math

p = 0.5   #获胜的概率
a = 2   #获胜的净回报
b = 1  #失败的净损失

randomNum = 10000 #随机样本数

randomSample = [1 if np.random.random_sample() < p else 0 for i in range(randomNum)] #

def isWin(i):
    global randomSample
    return randomSample[i] == 1

#根据样本模拟投机， 并计算所有投机完成后的最后资产
def calc_return(p, a, b, position, times):
    result = 1
    for i in range(times):
        #根据样本计算赌博之后的资产
        result = result * (1 - position) + result * position * ((1 + a) if isWin(i) else (1 - b))
        
    return result

#返回字典数组， [{仓位: postion， 回报率: rate},]
def monte_carlo_calc(p, a, b):
    
    result = []
    
    #穷举仓位， 并计算对应的回报率
    for x in range(101):
        position = x * 0.01
        total = calc_return(p, a, b, position, randomNum)
        rate = math.pow(total, 1.0/randomNum) - 1
        # print "仓位 = %.2f 复利 = %.3f" % (position, d)
        result.append({"position" : position, "rate" : rate})
        
    return result


results = monte_carlo_calc(p, a, b)

maxRate = -1
bestPostion = 0

for result in results:
    print "仓位 = %.2f 复利 = %.3f" % (result["position"], result["rate"])
    if result["rate"] > maxRate:
        maxRate = result["rate"]
        bestPostion = result["position"]

print "\r\n最佳仓位 = %.2f%% 复利 = %.2f%%" % (bestPostion*100, maxRate*100)
```

通过模拟可以大概确定最佳的下注为25%，复利大概为5.98%。 蒙特·卡罗方法的求解是一个模糊求解， 并不是确定性解。 这个案例里面对应的确定性解就是凯利公式了。当然，理论上说，当模拟的随机抽样足够大的时候，这个解就是精确解了。 感兴趣的朋友可以用这个方法求解圆周率，几行代码可以实现圆周率历史上千年的发展才能取得的成果。^ ^ 

在计算机时代蒙特·卡罗方法是一种很通用的解题方式。 当不知道确定性公式的时候（很多时候）， 我们可以绕开这个公式，得到一个（模糊的）答案。。。
