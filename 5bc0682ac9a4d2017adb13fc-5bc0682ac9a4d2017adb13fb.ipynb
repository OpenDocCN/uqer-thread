## 策略简介

ETF二八轮动已经为很多投资者所知，但简单根据20日动量进行ETF轮动的投资策略（即ETF轮动1.0版）存在一些问题：
1. 回撤过大
2. 夏普比率不高
3. 在震荡市表现不佳

基于此，我们对ETF轮动1.0版进行了一些优化的尝试。
1. 针对于回撤过大的问题，尝试设置固定的止损比例，每日扫描是否需要止损（v1.1）
2. 增加估值、波动率等因子，进行多因子轮动（v1.2）
3. 行业指数轮动（v1.3）

### 主要结论
1. 止损策略没有达到预期效果
2. 后续尝试多因子模型
