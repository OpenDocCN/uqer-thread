```python
import numpy as np
import matplotlib.pyplot as plot
import pandas as pd    

# import seaborn  as sns
# import matplotlib.pyplot as plt
# sns.set_style('white')
    
# 先取HS300股票 2010-01-01 到2017-12-31的每个月月末的因子值
#   先取交易日历
calendar_frame=DataAPI.TradeCalGet(exchangeCD=u"XSHG,XSHE",beginDate=u"2010-01-01",endDate=u"2017-12-31",field=u"",pandas="1")
hs300_frame=DataAPI.IdxConsGet(secID=u"",ticker=u"000300",isNew=u"",field=u"",pandas="1")
hs300_frame[hs300_frame.isNew==1]
tickers=list(hs300_frame.consTickerSymbol.values)
# 只取月末的日期
calendar_frame=calendar_frame[calendar_frame.isMonthEnd==1]
# 得到每一个月的最后一个交易日是几号
calendar_frame[:100]

factor_frame_list=[]
#   遍历每一天，拿到因子的数据
for tdate in calendar_frame.calendarDate.values:
    factor_frame=DataAPI.MktStockFactorsOneDayGet(
        tradeDate=tdate,ticker=tickers,field= ['secID','tradeDate','ROE','DAVOL10'],pandas="1")
    factor_frame_list.append(factor_frame)
# print(factor_frame_list)

#合并所有天的因子值
tfactor_frame=pd.concat(factor_frame_list)
tfactor_frame['tradeDate']=tfactor_frame['tradeDate'].apply(lambda x:x.replace('-',''))
tfactor_frame[:4] 

 # 拿到了hs300  所有股票 在这段时间这个月末的-----     对应的因子值


import time
from CAL.PyCAL import *


start_time=time.time()
# ------------------------------   设定回测参数------
start = '2010-01-05'                      # 回测起始时间
end = '2017-12-31'                       # 回测结束时间
benchmark = 'HS300'                       # 策略参考标准
universe = DynamicUniverse('HS300')      # 证券池，支持股票和基金
capital_base = 10000000                      # 起始资金
freq = 'd'                             # 策略类型，'d'表示日间策略使用日线回测，'m'表示日内策略使用分钟线回测
refresh_rate = Monthly(1)
accounts = {
    'fantasy_account': AccountConfig(account_type='security', capital_base=10000000)
}
  
# ------------------------------读取因子的数据-----------------
#  定义因子  重新命名 （以后研究别的也行）  重新设置索引   去掉空值
factor_date=tfactor_frame[['secID','tradeDate','ROE']]
factor_date.rename(columns={'ROE':'factor'}, inplace=True)
factor_date=factor_date.set_index('tradeDate',drop=True)
factor_date.dropna(inplace=True)
q_dates=factor_date.index.values
#  将里面的日期进行存起来
# print(factor_date)




# -----------------------------   将因子进行分组并且回测------------

#  调用quick_backtest方法
#   查看api quick backtest
sim_params=quartz.SimulationParameters(start,end,benchmark,universe,capital_base
                                     , refresh_rate=refresh_rate,accounts=accounts)
# 获取回测的行情数据
data=quartz.get_backtest_data(sim_params)
# 运行结果
result={}

# 遍历每一组，进行回测
for quantile_five in range(1,6):
    def initialize(context):                       # 初始虚拟账户
        pass
  

    def handle_data(context):
        # 定义 账户  和股票池   前一个交易的交易日期
        account=context.get_account('fantasy_account')
        current_universe=context.get_universe('stock',exclude_halt=True)
        pre_date=context.previous_date.strftime('%Y%m%d')
        if pre_date not in q_dates:
            return
        
        
        # 拿取调仓钱前一个交易日的因子，并按照相应的分位选择股票
        q=factor_date.ix[pre_date].dropna()
        q=q.set_index('secID',drop=True)
        q=q.ix[current_universe]
        q.dropna(inplace=True)
        q_min = q['factor'].quantile((quantile_five-1)*0.2)
        q_max = q['factor'].quantile(quantile_five*0.2)
        my_univ=q[(q['factor']>q_min) & (q['factor']<q_max)].index.values
        # 交易部分
        
        Position=account.get_positions()
        sell_list=[stk for stk in Position if stk not in my_univ]
        for stk in sell_list:
            account.order_to(stk,0)
        
        # 在目标股票池中 等权买入
        for stk in my_univ:
            account.order_pct(stk, 1.0/len(my_univ))
        
        
    # 生成策略对象
    strategy=quartz.TradingStrategy(initialize,handle_data)
    
    # 开始回测，返回说明见 
    bt,perf=quartz.quick_backtest(sim_params,strategy,data=data)
    
    # 保存运行结果，1 为因子最强组，5为因子的最弱组
    results[6-quantile_five]={'max_drawdown':perf['max_drawdown'],
                              'sharpe':perf['sharpe'],
                              'alpha':perf['alpha'],
                              'information_ratio':perf['information_ratio'],
                              'annualized_return':perf['annualized_return'],
                              'bt':bt}
    print('backtesting for group %s...............'% str(quantile_five)),
 
        
# ----------------- 进行画图展示----------------------

import seaborn  as sns
import matplotlib.pyplot as plt
sns.set_style('white')

fig=plt.figure(figsize=(10,8))
fig.set_tight_layout(True)
ax1=fig.add._subplot(211)
ax2=fig.add._subplot(212)
ax1.grid()
ax2.grid()




for qt in results:
    bt=results[qt]['bt']
    
    data=bt[[u'tradeDate',u'portfolio_value',u'benchmark_return']]
    data['portfolio_return']=data.portfolio_value/data.portfolio_value.shift(1)-1.0
    data['portfolio_return'].ix[0]=data['portfolio_return'].ix[0]/10000000-1.0
    data['excess_return']=data.portfolio_return-data.benchmark_return
    data['excess']=data.excess_return+1.0
    data['excess']=data.excess.cumprod()
    data['portfolio']=data.portfolio_return+1.0
    data['portfolio']=data.portfolio.cumprod()
    data['benchmark']=data.benchmark_return+1.0
    data['benchmark']=data.benchmark.cumprod()
    result[qt]['hedged_max_drawdown']=max([1-v/max(1,max(data['excess'][:i+1])) for i
                                         ,v in enumerate(data['excess'])])
    result[qt]['hedged_volatility']=np.std(data['excess'])*np.sqrt(252)
    result[qt]['hedged_annualized_return']=(data['excess'].values[-1])**(252.0/len(data['excess']))
    ax1.plot(data['tradeDate'],data[['portfolio']],label=str(qt))
    ax2.plot(data['tradeDate'],data[['excess']],label=str(qt))
    
ax1.legend(loc=0)
ax2.legend(loc=0)
ax1.set_ylabel(u"净值", fontproperties=font, fontsize=16)
ax2.set_ylabel(u'对冲净值',fontproperties=font,fontsize=16)
ax1.set_title(u'因子分组净值走势',fontproperties=font,fontsize=16)
ax2.set_title(u'因子对冲分组净值走势',fontproperties=font,fontsize=16)
#   这里的符号为什么要中文？
# ax1.set_ylabel(u"净值", fontproperties=font, fontsize=16)
# ax2.set_ylabel(u"对冲指数净值", fontproperties=font, fontsize=16)
# ax1.set_title(u"因子选股的多头组合", fontproperties=font, fontsize=16)

# results 转换成DF
results_pd=pd.DataFrame(results).T.sort_index()

results_pd=results_pd[[u'alpha',u'Beta',u'information_ratio', u'sharpe',
                      u'annualized_return',u'max_drawdown',u'hedged_max_drawdown',
                      u'hedged_volatility',u'hedged_annualized_return']]

cols=[(u'风险指标',u'Alpha'),
      (u'风险指标',u'Beta'),
      (u'风险指标',u'信息比率'),
    ( u'风险指标',u'夏普比率'),
     (u'纯股票多头',u'年华收益'),
     (u'纯股票多头',u'最大回撤'),
    ( u'对冲后',u'年化收益'),
     (u'对冲后',u'最大回撤'),
    ( u'对冲后',u'收益波动率')]


results_pd.columns=pd.MultiIndex.from_tuples(cols)
results_pd.index.name=u'五分位组别'
results_pd


    
    
    
    










```
