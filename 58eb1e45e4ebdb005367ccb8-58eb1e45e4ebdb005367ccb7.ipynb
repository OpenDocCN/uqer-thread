# 寻找有效因子

以下分享各因子按照最高最低1/10排列，每5天进行一次调仓，来测算近一段时期的有效因子情况。测算时间比较长，请耐心等待。

```python
import matplotlib.pylab as plt
import pandas as pd
pd.set_option('display.precision',2)
import pandas as pd
import numpy as np
from pandas import DataFrame,Series
from CAL.PyCAL import *
import matplotlib.pyplot as plt
from datetime import datetime
import seaborn as sns
import time
import datetime as dt
font.set_size(22)
small_font = font.copy()
small_font.set_size(12)

cal = Calendar('China.SSE')
today = datetime.today()
yesterday = cal.advanceDate(today, '-1B', BizDayConvention.Following)
week = cal.advanceDate(today, '-5B', BizDayConvention.Following)
month = cal.advanceDate(today, '-20B', BizDayConvention.Following)
season = cal.advanceDate(today, '-60B', BizDayConvention.Following)
halfyear = cal.advanceDate(today, '-120B', BizDayConvention.Following)
year = cal.advanceDate(today, '-250B', BizDayConvention.Following)
```

```python
factor_dic={'OperatingRevenueGrowRate':'每股营收同比增速','NetProfitGrowRate':'每股收益同比增速','FEARNG':'一致预期净利润增速','PB':'PB','PE':'PE','VOL20':'20日平均换手率','VOL60':'近60日日均换手率','VOL240':'近240日日均换手率','LFLO':'流通市值的对数','HBETA':'近250日CAMP模型的beta','RSTR12':'12月相对强势','get_new':'次新股','manager':'管理层20日增持'}
```

```python
universe1=set_universe('A')
universe1 = [stk[0:6] for stk in universe1]
def new_remove(ticker,tradeDate= None,day = 30):
    """
    给定股票列表,去除某个日期前多少个交易日,之后上市的新股
    Args:
        ticker (list of str): 需要进行筛选的股票列表（无后缀）
        tradeDate (datetime): 进行筛选的日期,默认为调用当天
        day (int): 向前漂移的交易日的个数
    Returns:
        list: 去掉新股股票之后的股票列表（无后缀）

    Examples:
        >> universe = set_universe('A')
        >> universe = [stk[0:6] for stk in universe]
        >> universe_without_new = new_remove(universe)
    """
    tradeDate = tradeDate if tradeDate is not None else datetime.datetime.now()
    period = '-' + str(day) + 'B'
    pastDate = cal.advanceDate(tradeDate,period)
    pastDate = pastDate.strftime("%Y-%m-%d")
    ipo_date = DataAPI.SecIDGet(partyID=u"",assetClass=u"e",ticker=ticker,cnSpell=u"",field=u"ticker,listDate",pandas="1")
    remove_list = ipo_date[ipo_date['listDate'] > pastDate]['ticker'].tolist()
    return [stk for stk in ticker if stk not in remove_list]

#需要做一个指定股票列表，指定日期后5天的平均涨幅
def average_yield(stocklist,day):
    df=pd.DataFrame()
    fiveday=cal.advanceDate(day, '5B', BizDayConvention.Following)
    
    price=DataAPI.MktEqudGet(tradeDate=day,ticker=stocklist,field=u"secID,closePrice",pandas="1")
    five_day_price=DataAPI.MktEqudGet(tradeDate=fiveday,ticker=stocklist,field=u"secID,closePrice",pandas="1")
    five_day_price.columns=['secID','5closePrice']
    price=pd.merge(price,five_day_price)
    price['average']=price['5closePrice']/price['closePrice']
    
    average_yield=price['average'].mean()
    return average_yield


def factor_stocklist_small(factor,day):
    #选出某日因子值中最小的十分之一，返回的是stocklist
    universe=new_remove(universe1,day)
    
    df=DataAPI.MktStockFactorsOneDayGet(tradeDate=day,ticker=universe,field=["ticker",factor],pandas="1").set_index('ticker')
    fdate = Date.parseISO(day)
   
    ep = df[factor].dropna().to_dict()
    ep = standardize(neutralize(winsorize(ep),fdate.strftime('%Y%m%d')))#中性化、去极化、标准化
    ep=pd.DataFrame(ep.items(), columns=['ticker', factor])
    df=ep.dropna()
        
        
    df=df.sort_values(by=factor)
    stklist=df['ticker'].tolist()
    num=len(stklist)
    num=num/10
    stklist=stklist[0:num]
    return stklist

def factor_stocklist_big(factor,day):
    #选出某日因子值中最大的十分之一，返回的是stocklist
    universe=new_remove(universe1,day)
    
    df=DataAPI.MktStockFactorsOneDayGet(tradeDate=day,ticker=universe,field=["ticker",factor],pandas="1").set_index('ticker')
    fdate = Date.parseISO(day)
   
    ep = df[factor].dropna().to_dict()
    ep = standardize(neutralize(winsorize(ep),fdate.strftime('%Y%m%d')))#中性化、去极化、标准化
    ep=pd.DataFrame(ep.items(), columns=['ticker', factor])
    df=ep.dropna()
    
    df=df.sort_values(by=factor)
    stklist=df['ticker'].tolist()
    num=len(stklist)
    num=num/10
    stklist=stklist[-num:-1]
    return stklist

#下面需要指定一个factor，日期倍数(5的倍数），如一年就输入50，每5日调仓，选择前最小的十分之一的股票，输出最后的收益率
def factor_profit_small(factor,n):
    #print yesterday
    profit=1
    for i in range(1,n+1):
        day=cal.advanceDate(yesterday, str(-5*i)+'B', BizDayConvention.Following)
        stocklist=factor_stocklist_small(factor,day)
        yield_tmp=average_yield(stocklist,day)
        profit=profit*yield_tmp
    return np.round((profit-1)*100,1)

#下面需要指定一个factor，日期倍数(5的倍数），如一年就输入50，每5日调仓，选择前最大的十分之一的股票，输出最后的收益率
def factor_profit_big(factor,n):
    #print yesterday
    profit=1
    for i in range(1,n+1):
        day=cal.advanceDate(yesterday, str(-5*i)+'B', BizDayConvention.Following)
        stocklist=factor_stocklist_big(factor,day)
        yield_tmp=average_yield(stocklist,day)
        profit=profit*yield_tmp
    return np.round((profit-1)*100,1)

#现在计算输入factor，给定一个日期列表（即一年，半年……），得出一个dataframe，行名是低factor和高factor，列名是时间段，内容是相应的收益率
def factor_profit_df(factor):
    days=[49,24,12,4,2,1]
    columns=['一年','半年','一季','一月','半月','一周']
    factor_name=factor_dic[factor]
    df=pd.DataFrame(index=['低'+factor_name,'高'+factor_name])
    for i in days:
        profit_big=factor_profit_big(factor,i)
        profit_small=factor_profit_small(factor,i)
        df.loc['低'+factor_name,i]=profit_small
        df.loc['高'+factor_name,i]=profit_big
    df.columns=columns
    return df.T

#自定义高管增持factor
def manager(day):
    begindate=cal.advanceDate(day, '-20B', BizDayConvention.Following)
    df=DataAPI.EquMsChangesGet(ticker=universe1,beginDate=begindate,endDate=day,field=u"ticker,changeProp",pandas="1").dropna()
    df=df['changeProp'].groupby(df['ticker']).sum().sort_values(ascending=False)
    stklist=df[0:50].index.tolist()
    return stklist

#自定义一月反转factor
def one_month_reverse(day):
    begindate=cal.advanceDate(day, '-20B', BizDayConvention.Following)
    df=DataAPI.EquMsChangesGet(ticker=universe1,beginDate=begindate,endDate=day,field=u"ticker,changeProp",pandas="1").dropna()
    df=df['changeProp'].groupby(df['ticker']).sum().sort_values(ascending=False)
    stklist=df[0:50].index.tolist()
    return stklist

#自定义次新股factor
def get_new(tradeDate):
    """
    给定股票列表,获得某个日期前多少个交易日到这个日期之间上市的新股
    注：可以定义指定日前30日至90日上市者为次新股
    Args:
        ticker (list of str): 需要进行筛选的股票列表（无后缀）
        tradeDate (datetime): 进行筛选的日期,默认为调用当天
        day (int): 向前漂移的交易日的个数
    Returns:
        list: 获取新股的股票列表（无后缀）

    Examples:
        >> universe = set_universe('A')
        >> universe = secID2ticker(universe)
        >> new_stock = get_new(universe)
    """
    tradeDate = tradeDate if tradeDate is not None else datetime.datetime.now()
    period1 = '-30B'
    period2='-90B'
    pastDate1 = cal.advanceDate(tradeDate,period1)
    pastDate1 = pastDate1.strftime("%Y-%m-%d")
    pastDate2 = cal.advanceDate(tradeDate,period2)
    pastDate2 = pastDate2.strftime("%Y-%m-%d")
    ipo_date = DataAPI.SecIDGet(partyID=u"",assetClass=u"e",ticker=universe1,cnSpell=u"",field=u"ticker,listDate",pandas="1")
    new_list = ipo_date[(ipo_date['listDate'] > pastDate2) & (ipo_date['listDate'] < pastDate1)]['ticker'].tolist()
    return new_list

#下面针对自定义factor，输出最后的收益率
def stklist_profit(factor,n):
    #print yesterday
    profit=1
    for i in range(1,n+1):
        day=cal.advanceDate(yesterday, str(-5*i)+'B', BizDayConvention.Following)
        stocklist=factor(day)
        yield_tmp=average_yield(stocklist,day)
        profit=profit*yield_tmp
    return np.round((profit-1)*100,1)

#现在计算输入自定义factor，给定一个日期列表（即一年，半年……），得出一个dataframe，行名是'factor名称'，列名是时间段，内容是相应的收益率
def factor_me_profit_df(factor):
    days=[49,24,12,4,2,1]
    columns=['一年','半年','一季','一月','半月','一周']
    factor_name=factor_dic[factor.__name__]
    print factor_name
    df=pd.DataFrame(index=[factor_name])
    for i in days:
        profit=stklist_profit(factor,i)
        df.loc[factor_name,i]=profit
    df.columns=columns
    return df.T

#最终表格，其中factors1为系统自带，factors2为自定义
def factor_profit_all(factors1,factors2):
    secs_time = 0
    start_time = time.time()
    df=pd.DataFrame()
    n=len(factors1)+len(factors2)
    print 'There are '+str(n)+ ' factors.'
    i=1
    for factor in factors1:
        tmp=factor_profit_df(factor)
        if df.empty:
            df=tmp
        else:
            df=pd.merge(df,tmp,left_index=True,right_index=True)
        print str(i) + ' of '+ str(n) +' factor '+str(factor)+' is done.'
        i=i+1
        finish_time = time.time()#结束时间
        print '  ' + str(np.round((finish_time-start_time) - secs_time, 0)) + ' seconds elapsed.'#耗时
        secs_time = (finish_time-start_time)#调整已用时间
        
    for factor in factors2:
        tmp=factor_me_profit_df(factor)
        if df.empty:
            df=tmp
        else:
            df=pd.merge(df,tmp,left_index=True,right_index=True)
        print str(i) + ' of '+ str(n) +' factor '+ factor.__name__ +' is done.'
        i=i+1
        finish_time = time.time()#结束时间
        print '  ' + str(np.round((finish_time-start_time) - secs_time, 0)) + ' seconds elapsed.'#耗时
        secs_time = (finish_time-start_time)#调整已用时间    
    return df
```
