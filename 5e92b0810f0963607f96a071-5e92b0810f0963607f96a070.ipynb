```python
import numpy as np
import pandas as pd
'''
——————————————————————————————————————————————————
                    交易日
——————————————————————————————————————————————————
'''
date_list=DataAPI.TradeCalGet(exchangeCD=u"XSHG",beginDate=u"",endDate=u"",field=u"",pandas="1")
date_list = date_list[date_list.isOpen==1].calendarDate.tolist()
import datetime
now = str(datetime.date.today())
date_list=list(filter(lambda x:x<=now,date_list))
date_list=list(filter(lambda x:x>'2019-01-01',date_list))
start = '2019-01-01'
end = '2020-02-25'
df = pd.read_excel('ETF基金列表.xlsx',sheet_name='Sheet1')
ETF_list = df.iloc[:,0].values.tolist()    # 268个
```

```python
'''——————————————————————————获取基金基本信息，导出股票型、混合型基金————————————————————————————————'''
df=DataAPI.FundGet(secID=u"",ticker=u"",etfLof=u"",listStatusCd=u"",category=u"",idxID=u"",idxTicker=u"",operationMode=u"",beginDate="",endDate="",status="",field=u"",pandas="1")
df = df[df.category.isin(['E','H'])]
df = df.drop(['secID','investField','investTarget'],axis=1)
ticker_list = df[~(df.expireDate<'2018-12-31')].ticker.values.tolist()
df.to_csv('基金基本信息.csv',encoding='gbk')
```

```python
# 下载主动型基金的季报、半年报、年报的持仓等数据
import pandas as pd
df_all = pd.DataFrame()
for i,ticker in enumerate(ticker_list):
    if i%100==0:print(i)
    '''—————————————————————————————————— 主动型基金 ——————————————————————————————————————'''
    df1 = DataAPI.FundHoldingsGet(secID=u"",ticker=ticker,reportDate=u"",beginDate=u"2018-12-12",endDate=u"",secType="",holdingTicker="",holdingSecID="",field=['ticker','reportDate','holdingsecType','holdingTicker','holdVolume','marketValue','ratioInNa','publishDate'],pandas="1").sort_values(['reportDate','ratioInNa'],ascending=[True,False])
    df1=df1[df1.holdingsecType=='E']
    df1=df1.drop(['holdingsecType'],axis=1)
    df_all = df_all.append(df1)
df_all = df_all.append(df1)
df_all.to_hdf('func_holding.h5', key='df', mode='w', complevel=9, complib='blosc')
```
