# FCFP(Free Cash Flow to Price)

## **简介**
-----------

计算并测试了自由现金流/市值(FCFP)因子，并比较了各种因子构建方法的优劣。

在全A股近八年的回测中，根据该因子构建的多头组合有1.85的IR、27%的年化收益率和56%的日度胜率

## **因子逻辑与定义**

估值效应是一个广泛并且长期存在的市场异象，这在各种经典论文中均有阐述，传统的BP、ETOP等因子正是利用这一异象的代表。然而随着套利者的增加，传统估值因子的效果正在不断弱化，如何才能在这一背景下更好地利用估值效应呢？一个自然的想法就是构建能够更好反映公司价值的指标。

**自由现金流(Free Cash Flow, FCF)**作为代表公司可以自由支配现金数量的会计变量，具有很多良好的性质。
- 由于该指标的信息来自于现金流量表，相较于权责发生制下的会计收益，FCF很难被操纵和伪造
- FCF从现金流的角度更好地衡量了一个公司的盈利能力

基于以上原因，类比EP、CFP等估值因子的结构，构建基于FCF的估值因子：***FCFP***

我们知道，自由现金流的定义为经营性净现金流减去资本支出，而FCFP为自由现金流除总市值，若上下同时除总股本，则变成$FCFP = \frac{每股自由现金流}{每股股价} = \frac{FCFPS}{ClosePrice}$

接下来我们将要面临两个问题：
1. 我们该用哪种指标来代表每股自由现金流？

2. 对于银行和非银金融行业的自由现金流如何计算？如何处理？

对于问题1，优矿的因子DataAPI（专业版）中提供两种每股自由现金流，分别是`每股企业自由现金流`EnterpriseFCFPS和`每股股东自由现金流`ShareholderFCFPS，主要区别在于企业自由现金流并未考虑期内的债务增加额，我们分别考虑并计算。

其中，$EnterpriseFCFPS = \frac{经营活动现金流量净额-资本支出-净利息费用*税率}{总股数}$，$ShareholderFCFPS=\frac{经营活动现金流量净额-资本支出+债务增加额-净利息费用}{总股数}$

非专业版用户可以按照公式读取会计科目自行计算，在下文中我会直接使用优矿计算好的因子。

对于问题2，在EnterpriseFCFPS和ShareholderFCFPS的计算过程中，均没有考虑金融企业，所以银行等行业的值均为NaN，关于金融企业自由现金流的计算方法有很多[资料](http://barrons.blog.caixin.com/archives/2736)和学术论文，有兴趣可以深入研究；
在这里，我们对于金融企业有两种处理方式：一为直接dropna()，即最终选股结果中不包括这些企业；二为统一将这些企业的因子值记为0，后面会有相应的比较分析。

### Test 1: ShareholderFCFPS + 金融企业dropna()

```python
import pandas as pd
import numpy as np
import time
```

```python
def FCFP_get(universe, begin, end, file_name=None):
    """
    计算自由现金流总市值比因子
    
    universe: list, secID组成的list
    begin: datetime string, 起始日期，格式为"%Y%m%d"
    end: datetime string, 终止日期，格式为"%Y%m%d"
    file_name： string, 以".csv"结尾且符合文件命名规范的字符串
    """

    cal_dates = DataAPI.TradeCalGet(exchangeCD=u"XSHG", beginDate=begin, endDate=end).sort('calendarDate')
    cal_dates = cal_dates[cal_dates['isOpen']==1]
    all_dates = cal_dates['calendarDate'].values.tolist()
    all_dates = [x.replace('-', '') for x in all_dates]
    
    trigger_date = [x.replace('-', '') for x in all_dates]
    
    print 'Factor will be calculated for ' + str(len(trigger_date)) + ' days:'
    count = 0
    secs_time = 0
    start_time = time.time()
    
    ret_data = pd.DataFrame()
    ret = dict()
    for dt in trigger_date: 
        # 横截面因子计算逻辑
        for i in range(5):
            try:
                FCFPS_data = DataAPI.MktStockFactorsOneDayProGet(tradeDate=dt,secID=universe,field=u"secID,ShareholderFCFPS",pandas="1").set_index('secID')
                closePrice_data = DataAPI.MktEqudGet(tradeDate=dt,secID=universe,field=["secID", "closePrice"],pandas="1").set_index('secID')
                break
            except Exception, e:
                if i == 4:
                    import traceback
                    print traceback.format_exc()
                    raise e
        
        temp_df = pd.concat([FCFPS_data, closePrice_data], axis=1)
        ret[dt] = (temp_df['ShareholderFCFPS']/temp_df['closePrice']).dropna()  # 将NaN drop，因子值将不包含金融企业
        # print "date is ", dt_ret
        # 打印进度部分
        count += 1
        if count > 0 and count % 100 == 0:
            finish_time = time.time()
            print count,
            print '  ' + str(np.round((finish_time-start_time) - secs_time, 0)) + ' seconds elapsed.'
            secs_time = (finish_time-start_time)
            
    ret_data = pd.DataFrame(ret).T
    ret_data = ret_data.fillna(0)
    
    if file_name:
        ret_data.to_csv(file_name)
            
    return ret_data
```
