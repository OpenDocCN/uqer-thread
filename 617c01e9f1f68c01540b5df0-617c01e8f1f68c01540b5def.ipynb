# 无脑蒙特卡洛欧式/亚式期权定价
optionSolve() 为基于monte_carlo
参数含义
underlying_price=2.382             标的股票价格
strike_price=2.4                   期权执行价
volatility=0.2225                  年化波动率
call_put='call'                    期权的购沽类型 '认购'
residual_maturity=31               期权剩余的行权期
risk_free_rate=0.019               无风险利率
trading_days_of_year=252           年交易日数量
monte_carlo_times=50               蒙特卡洛次数
duringDays=2                       期权价值覆盖天数 若为 1 则为欧式 大于1 则为普通亚式
need_df=True                            最后一次模拟路径落地

```python
import pandas as pd
import numpy as np
import math
import datetime

underlying_price=2.382             
strike_price=2.4                   
volatility=0.2225                  
call_put='call'                    
residual_maturity=31                            
risk_free_rate=0.019               
trading_days_of_year=252           
monte_carlo_times=50              
during_days=2 
need_df=True
need_df=False

optionValueList = []
for i in range(monte_carlo_times):
    # 蒙特卡洛基础
    # 指标的（股票、利率、汇率、基金）价格的对数收益率服从（mu=0，sigma）的正态分布
    # 模拟对数收益率序列
    df = pd.DataFrame()
    df['ri'] = np.random.normal(loc=0, scale=volatility / (trading_days_of_year ** 0.5), size=residual_maturity)
    # 基于初始价格计算模拟后的价格
    df['close'] = [underlying_price for x in range(residual_maturity)]

    for indexI in range(residual_maturity):
        if indexI > 0:
            df.loc[indexI, 'close'] = df.loc[indexI - 1, 'close'] * math.e ** df.loc[indexI, 'ri']

    # print(df)
    # 计算 覆盖天数 内的 平均价格
    if during_days>0:
        dfNeed = df.loc[:,'close'].iloc[-during_days:]
        meanData = dfNeed.mean()
        # print(meanData)
        # print(dfNeed)

        # 根据类型计算行权收益的现值
        if call_put == 'call':
            exeValue = max(meanData-strike_price,0)
        else:
            exeValue = max(strike_price-meanData,0)

        exeValueP = exeValue*math.e**(-risk_free_rate*residual_maturity/trading_days_of_year)

    else:
        exeValueP = 0
    optionValueList.append(exeValueP)
    
print(optionValueList)

# 是否落地最后一次数据
if need_df:
    now = datetime.datetime.now().strftime('%Y%m%d %H%M%S')
    df.to_excel('\optionSolve_'+now+'.xlsx')
    
# 输出期权价值的期望值 作为 价格
print(sum(optionValueList)/len(optionValueList))
        
```
