```python
import DataAPI
import numpy as np
import pandas as pd
import talib as ta
from collections import deque
from itertools import product
import datetime
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import seaborn as sns
import quartz_futures as qf
from quartz_futures.api import *
from CAL import *
cal = Calendar('China.SSE') #上海证券交易所交易日历
#保证建仓日 #exchangeCD:上交所
cal_dates = DataAPI.TradeCalGet(exchangeCD=u"XSGE", beginDate='20140101', endDate='20161229').sort('calendarDate')#获取上海证券交易所，开始日期从20140101，结束日期=20161229，calendardate：日期，isOpen：日期当天是否开市，0表否，1表是
cal_dates = cal_dates[cal_dates['isOpen']==1] #剔除isopen中为0的列
trade_list = cal_dates['calendarDate'].values.tolist() ## 将某一列转化成list

pingzhong = ['TF1709', 'TF1712', 'TF1803', 'T1709', 'T1712', 'T1803', 'HTFF1709', 'HTFF1712']
arrays = [[u'CFFEX五年期', u'CFFEX五年期', u'CFFEX五年期', u'CFFEX十年期', u'CFFEX十年期', u'CFFEX十年期', u'HKEX五年期', u'HKEX五年期'], pingzhong]  #第一列按照所写结果呈现，第二列按照pingzhong_chinese所写内容呈现
tuples = list(zip(*arrays))
index = pd.MultiIndex.from_tuples(tuples, names=[u'类别', u'品种'])
turnoverValue = [] #成交金额
turnoverVol = [] #成交数量，国内单位为手
on_market_date = []
trade_multi = []
minChgPriceNum = []
tradeMarginRatio = []
for j in range(len(pingzhong)):
    data = DataAPI.MktMFutdGet(startDate=trade_list[0], endDate=trade_list[-1],  contractObject=pingzhong[j], field=[u"turnoverValue", u"turnoverVol"], mainCon=1,pandas="1") #获取四大期货交易所的日行情 #contractobject:期货合约标的 #mainCon：是否主力
    turnoverValue.append(data['turnoverValue'].tolist())
    turnoverVol.append(data['turnoverVol'].tolist())
    date = DataAPI.FutuGet(contractStatus="DE",contractObject=pingzhong[j],field=[u"listDate"],pandas="1") #获取期货合约信息 #期货合约状态 DE终止上市 L上市 #pandas=1 返回dataframe
    on_market_date.append(date.ix[0,'listDate'])
    trade_detail = DataAPI.FutuGet(contractStatus="L",contractObject=pingzhong[j],
                                   field=[u"contMultUnit", u"contMultNum", u"minChgPriceNum", u"minChgPriceUnit", u"tradeMarginRatio"], pandas="1")
    trade_multi.append(str(trade_detail.ix[0, u"contMultNum"])+trade_detail.ix[0, u"contMultUnit"])
    minChgPriceNum.append(str(trade_detail.ix[0, u"minChgPriceNum"])+trade_detail.ix[0, u"minChgPriceUnit"])
    tradeMarginRatio.append("%.2f%%" % (trade_detail.ix[0, u"tradeMarginRatio"]))
    
avg_turnoverValue = []
avg_turnoverVol = []
for i in range(len(pingzhong)):
    avg_turnoverValue.append(np.mean(turnoverValue[i]))
    avg_turnoverVol.append(np.mean(turnoverVol[i]))

table1 = pd.DataFrame(columns=[u'代码', u'2014年以来日均成交额', u'2014年以来日均成交量', u'上市日期', u'交易单位', u'最小变动价位', u'最低交易保证金比例'], index=index)
table1[u'代码'] = pingzhong
table1[u'2014年以来日均成交额'] = avg_turnoverValue
table1[u'2014年以来日均成交量'] = avg_turnoverVol
table1[u'上市日期'] = on_market_date
table1[u'交易单位'] = trade_multi
table1[u'最小变动价位'] = minChgPriceNum  
table1[u'最低交易保证金比例'] = tradeMarginRatio
table1
```
