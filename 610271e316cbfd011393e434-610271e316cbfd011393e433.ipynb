```python
"""
初始化内容
"""
import pandas as pd
import statsmodels.api as sm
import numpy as np
import datetime
startDate = '20200101'
endDate = '20210701'
trade_day = DataAPI.TradeCalGet(exchangeCD=u"XSHG",beginDate=startDate,endDate=endDate,isOpen=u"1",field=u"",pandas="1")['calendarDate']
fundPool = ['000689', '000690']
industryPool = ['801010',
 '801020',
 '801030',
 '801040',
 '801050',
 '801080',
 '801110',
 '801120',
 '801130',
 '801140',
 '801150',
 '801160',
 '801170',
 '801180',
 '801200',
 '801210',
 '801230',
 '801710',
 '801720',
 '801730',
 '801740',
 '801750',
 '801760',
 '801770',
 '801780',
 '801790',
 '801880',
 '801890']
```

```python
# 记录所有基金的return
fund_info = {}
for index in fundPool: 
    fundReturn = DataAPI.FundNavGet(secID=u"",ticker=index,dataDate=u"",beginDate=startDate,endDate=endDate,partyID="",partyShortName="",field=u"",pandas="1")[['endDate','secShortName', 'adjNavChgPct']]
    fund_info[index] = fundReturn
```

```python
# 行业的return
industry_info = {}
for ind in industryPool:
    indReturn = DataAPI.MktIdxdGet(indexID=u"",ticker=ind,tradeDate=u"",beginDate=u"20200101",endDate=u"20210701",exchangeCD=u"XSHE,XSHG",field=u"",pandas="1")[['tradeDate', 'secShortName', 'indexID', 'CHGPct']]
    indReturn['CHGPct'] = indReturn['CHGPct'] * 100
    industry_info[ind] = indReturn
```

```python
# 国家return
countryReturn = DataAPI.MktIdxdGet(indexID=u"",ticker='000905',tradeDate=u"",beginDate=u"20200101",endDate=u"20210701",exchangeCD=u"XSHE,XSHG",field=u"",pandas="1")[['tradeDate', 'secShortName', 'indexID', 'CHGPct']]
countryReturn['CHGPct'] = countryReturn['CHGPct'] * 100
```

```python
# 获取beta因子信息
beta_return_list = []
for day in trade_day: 
    print(day)
    BETA = DataAPI.MktStockFactorsOneDayGet(tradeDate=day,secID=u"",ticker=u"",field=u"",pandas="1")[['secID', 'tradeDate', 'HBETA']].dropna()
    beta_len = BETA.shape[0]
    beta_30per = int(round(beta_len*0.3))
    BETA_top30_stock = BETA.sort_values(by='HBETA', ascending=False)[:beta_30per]['secID']
    BETA_last30_stock = BETA.sort_values(by='HBETA', ascending=False)[-beta_30per:]['secID']
    top30_return = DataAPI.EquRetudGet(secID=BETA_top30_stock,
                                        ticker=u"",listStatusCD=u"",beginDate=day,endDate=day,dailyReturnNoReinvLower=u"",
                                        dailyReturnNoReinvUpper=u"",dailyReturnReinvLower=u"",
                                        dailyReturnReinvUpper=u"",isChgPctl=u"",field=u"dailyReturnReinv",pandas="1").mean()
    last30_return = DataAPI.EquRetudGet(secID=BETA_last30_stock,
                                        ticker=u"",listStatusCD=u"",beginDate=day,endDate=day,dailyReturnNoReinvLower=u"",
                                        dailyReturnNoReinvUpper=u"",dailyReturnReinvLower=u"",
                                        dailyReturnReinvUpper=u"",isChgPctl=u"",field=u"dailyReturnReinv",pandas="1").mean()
    beta_return = top30_return - last30_return
    beta_return_list.append(beta_return[0])
    
beta_return_ser = pd.Series(beta_return_list)
beta_return_df = pd.DataFrame({'TradeDate': trade_day, 'BetaReturn': beta_return_ser})
beta_return_df.to_excel("BetaReturn.xlsx")
```

```python
# 获取动量因子信息
mom_return_list = []
for day in trade_day: 
    print(day)
    MOMENTUM = DataAPI.MktStockFactorsOneDayGet(tradeDate=day,secID=u"",ticker=u"",field=u"",pandas="1")[['secID', 'tradeDate', 'RSTR24']].dropna()
    # 根据过去504个交易日收益率加权算得动量因子
    MOM_len = MOMENTUM.shape[0]
    MOM_30per = int(round(MOM_len*0.3))
    MOM_top30_stock = MOMENTUM.sort_values(by='RSTR24', ascending=False)[:MOM_30per]['secID']
    MOM_last30_stock = MOMENTUM.sort_values(by='RSTR24', ascending=False)[-MOM_30per:]['secID']
    top30_return = DataAPI.EquRetudGet(secID=MOM_top30_stock,
                                   ticker=u"",listStatusCD=u"",beginDate=day,endDate=day,dailyReturnNoReinvLower=u"",
                                   dailyReturnNoReinvUpper=u"",dailyReturnReinvLower=u"",
                                   dailyReturnReinvUpper=u"",isChgPctl=u"",field=u"dailyReturnReinv",pandas="1").mean()
    last30_return = DataAPI.EquRetudGet(secID=MOM_last30_stock,
                                        ticker=u"",listStatusCD=u"",beginDate=day,endDate=day,dailyReturnNoReinvLower=u"",
                                        dailyReturnNoReinvUpper=u"",dailyReturnReinvLower=u"",
                                        dailyReturnReinvUpper=u"",isChgPctl=u"",field=u"dailyReturnReinv",pandas="1").mean()
    mom_return = top30_return - last30_return
    mom_return_list.append(mom_return[0])
MOM_return_ser = pd.Series(mom_return_list)
MOM_return_df = pd.DataFrame({'TradeDate': trade_day, 'MomReturn': MOM_return_ser})
MOM_return_df.to_excel("MOMReturn.xlsx")
```

```python
# 获取残差波动因子信息
Vol_return_list = []
for day in trade_day: 
    print(day)
    Volatility = DataAPI.MktStockFactorsOneDayGet(tradeDate=day,secID=u"",ticker=u"",field=u"",pandas="1")[['secID', 'tradeDate', 'DVRAT', 'CMRA', 'HSIGMA']].dropna()
    Volatility['DASTD'] = 0.74*Volatility['DVRAT'] + 0.16*Volatility['CMRA'] + 0.1*Volatility['HSIGMA']
    # 根据过去504个交易日收益率加权算得动量因子
    Vol_len = Volatility.shape[0]
    Vol_30per = int(round(Vol_len*0.3))
    Vol_top30_stock = Volatility.sort_values(by='DASTD', ascending=False)[:Vol_30per]['secID']
    Vol_last30_stock = Volatility.sort_values(by='DASTD', ascending=False)[-Vol_30per:]['secID']
    top30_return = DataAPI.EquRetudGet(secID=Vol_top30_stock,
                                   ticker=u"",listStatusCD=u"",beginDate=day,endDate=day,dailyReturnNoReinvLower=u"",
                                   dailyReturnNoReinvUpper=u"",dailyReturnReinvLower=u"",
                                   dailyReturnReinvUpper=u"",isChgPctl=u"",field=u"dailyReturnReinv",pandas="1").mean()
    last30_return = DataAPI.EquRetudGet(secID=Vol_last30_stock,
                                        ticker=u"",listStatusCD=u"",beginDate=day,endDate=day,dailyReturnNoReinvLower=u"",
                                        dailyReturnNoReinvUpper=u"",dailyReturnReinvLower=u"",
                                        dailyReturnReinvUpper=u"",isChgPctl=u"",field=u"dailyReturnReinv",pandas="1").mean()
    Vol_return = top30_return - last30_return
    Vol_return_list.append(Vol_return[0])
Vol_return_ser = pd.Series(Vol_return_list)
Vol_return_df = pd.DataFrame({'TradeDate': trade_day, 'VolReturn': Vol_return_ser})
Vol_return_df.to_excel("VolReturn.xlsx")
```

```python
# 获取非线性因子，并计算非线性排名
NSize_return_list = []
for day in trade_day: 
    print(day)
    my_data = DataAPI.MktStockFactorsOneDayGet(tradeDate=day,secID=u"",ticker=u"",field=u"",pandas="1")[['secID', 'LCAP']].dropna()
    my_x = my_data['LCAP']
    my_y = my_x**3
    model = sm.OLS(my_y,my_x).fit()
    my_slope = model.params[0]
    my_data['non-linear_size_factor'] = my_data['LCAP'].apply(lambda x: x**3 - my_slope*x)
    my_data['abs-non-linear_size_factor'] = my_data['LCAP'].apply(lambda x: np.abs(x**3 - my_slope*x))
    my_data = my_data.sort_values(by='non-linear_size_factor', ascending=False)
    NSize_len = my_data.shape[0]
    NSize_10per = int(round(NSize_len*0.15))
    NSize_20per = int(round(NSize_len*0.15))
    NSize_top15_stock = my_data.sort_values(by='non-linear_size_factor', ascending=False)[:NSize_10per]['secID']
    NSize_last15_stock = my_data.sort_values(by='non-linear_size_factor', ascending=False)[-NSize_20per:]['secID']
    NSize_top30_stock = pd.concat([NSize_top15_stock,NSize_last15_stock])
    my_data = my_data.sort_values(by='abs-non-linear_size_factor', ascending=False)
    NSize_30per = int(round(NSize_len*0.3))
    NSize_last30_stock = my_data.sort_values(by='non-linear_size_factor', ascending=False)[-NSize_30per:]['secID']
    top30_return = DataAPI.EquRetudGet(secID=NSize_top30_stock,
                                   ticker=u"",listStatusCD=u"",beginDate=day,endDate=day,dailyReturnNoReinvLower=u"",
                                   dailyReturnNoReinvUpper=u"",dailyReturnReinvLower=u"",
                                   dailyReturnReinvUpper=u"",isChgPctl=u"",field=u"dailyReturnReinv",pandas="1").mean()
    last30_return = DataAPI.EquRetudGet(secID=NSize_last30_stock,
                                        ticker=u"",listStatusCD=u"",beginDate=day,endDate=day,dailyReturnNoReinvLower=u"",
                                        dailyReturnNoReinvUpper=u"",dailyReturnReinvLower=u"",
                                        dailyReturnReinvUpper=u"",isChgPctl=u"",field=u"dailyReturnReinv",pandas="1").mean()
    NSize_return = top30_return - last30_return
    NSize_return_list.append(NSize_return[0])
    print(NSize_return[0]*100)
# NSize_return_ser = pd.Series(NSize_return_list)
# NSize_return_df = pd.DataFrame({'TradeDate': trade_day, 'NSizeReturn': NSize_return_ser})
# NSize_return_df.to_excel("NSizeReturn.xlsx")

```

```python
# 获取流动性因子
Liq_return_list = []
for day in trade_day: 
    print(day)
    my_data = DataAPI.MktStockFactorsOneDayGet(tradeDate=day,secID=u"",ticker=u"",field=u"",pandas="1")[['secID', 'ticker', 'VOL20', 'VOL60', 'VOL240']].dropna()
    new_date = (datetime.datetime.strptime(day, '%Y-%m-%d') - datetime.timedelta(days=365)).strftime("%Y-%m-%d")  # 上市日期要在一年前
    data_use = DataAPI.SecIDGet(ticker=my_data['ticker'],assetClass='E',field=["secID","listDate"],pandas="1")
    data_use['判断'] = data_use['listDate'] < new_date
    data_use = data_use[data_use['判断']==True]
    data_use.index = data_use['secID']
    my_data.index = my_data['secID']
    my_data = pd.concat([data_use, my_data], axis=1, join='inner')
    my_data['VOL20'] = np.log(20*my_data['VOL20'])
    my_data['VOL60'] = np.log(60*my_data['VOL60'])
    my_data['VOL240'] = np.log(240*my_data['VOL240'])
    my_data['Liquidity'] = my_data.apply(lambda x: 0.35*x['VOL20'] + 0.35*x['VOL60'] + 0.3*x['VOL240'], axis=1)
    my_data = my_data.sort_values(by='Liquidity', ascending=False)
    my_len = my_data.shape[0]
    my_data_30per = int(round(my_len*0.3))
    my_data_top30_stock = my_data[:my_data_30per].index
    my_data_last30_stock = my_data[-my_data_30per:].index
    top30_return = DataAPI.EquRetudGet(secID=my_data_top30_stock,
                                   ticker=u"",listStatusCD=u"",beginDate=day,endDate=day,dailyReturnNoReinvLower=u"",
                                   dailyReturnNoReinvUpper=u"",dailyReturnReinvLower=u"",
                                   dailyReturnReinvUpper=u"",isChgPctl=u"",field=u"dailyReturnReinv",pandas="1").mean()
    last30_return = DataAPI.EquRetudGet(secID=my_data_last30_stock,
                                        ticker=u"",listStatusCD=u"",beginDate=day,endDate=day,dailyReturnNoReinvLower=u"",
                                        dailyReturnNoReinvUpper=u"",dailyReturnReinvLower=u"",
                                        dailyReturnReinvUpper=u"",isChgPctl=u"",field=u"dailyReturnReinv",pandas="1").mean()
    my_return = top30_return - last30_return
    print(my_return[0])
    Liq_return_list.append(my_return[0])
Liq_return_ser = pd.Series(Liq_return_list)
Liq_return_df = pd.DataFrame({'TradeDate': trade_day, 'LiqReturn': Liq_return_ser})
Liq_return_df.to_excel("LiqReturn.xlsx")
```
