```python
import pandas as pd
import numpy as np

# 定义每年5月最后一个交易日的分组函数，分为6组，目的是为得到因子
def get_6groups(breakpoint):            #breakpoint为str
        
    #获取每5月最后一个交易日期
    cal_date = DataAPI.TradeCalGet(exchangeCD=u"XSHG",beginDate=str(int(breakpoint)-20),endDate=breakpoint,\
                                    field=u"calendarDate,isOpen",pandas="1")
    cal_date = cal_date[cal_date['isOpen']==1]
    breakpoint = cal_date['calendarDate'].tolist()[-1]
    
    universe = set_universe('000906.ZICN',date=breakpoint)  #选取universe时候必须注明时间,取当时的中证800
    #每年531日，重新取中证800的成分股
    #对每5月最后一个交易日期的中证800按照市值分为S B 两组
    
    ME= DataAPI.MktEqudGet(secID=universe,ticker=u"",tradeDate=breakpoint,\
                            field=u"secID,ticker,marketValue",pandas="1").dropna()
    ME_50 = np.percentile(ME['marketValue'].tolist(),50)  #np.percentile内部会对其由小到大排序
    B = ME[ME['marketValue']>=ME_50]['secID'].tolist()
    S = ME[ME['marketValue']<ME_50]['secID'].tolist()
    #a = len(ME)-len(S)-len(B)
    
    #对每5月最后一个交易日期的中证800按照BP(PB大于0的倒数）分为H M L
    BP= DataAPI.MktStockFactorsOneDayGet(secID=universe,ticker=u"",tradeDate=breakpoint,\
                            field=u"secID,ticker,PB",pandas="1").dropna()
    BP=BP[BP['PB']>0]
    BP['BP']=1/BP['PB']
    BP=BP.drop('PB',axis = 1)
    BP_30 = np.percentile(BP['BP'].tolist(),30)
    BP_70 = np.percentile(BP['BP'].tolist(),70)
    H=BP[BP['BP']>=BP_70]['secID'].tolist()
    M=BP[(BP['BP']>=BP_30)&(BP['BP']<BP_70)]['secID'].tolist()
    L=BP[BP['BP']<BP_30]['secID'].tolist()

    #分别取两种分组的交集得到六个分组
    SH =list(set(S).intersection(set(H)))
    SM =list(set(S).intersection(set(M)))
    SL =list(set(S).intersection(set(L)))
    BH =list(set(B).intersection(set(H)))
    BM =list(set(B).intersection(set(M)))
    BL =list(set(B).intersection(set(L)))
    
    return SH,SM,SL,BH,BM,BL
```

```python
#定义获取某一组每年的月度收益率数据
def get_returnMonthly(x,year):
    
    returnMonthly = []                     # 用于存储12个月的组合加权收益率
    # 获取本年5月到下一年5月连续13个月的每月末交易日历
    cal_list = DataAPI.TradeCalGet(exchangeCD=u"XSHG",beginDate=str(int(year)*10000+515),\
                                   endDate=str(int(year+1)*10000+615),\
                        field=u"calendarDate,isMonthEnd",pandas="1")
    cal_list=cal_list[cal_list['isMonthEnd']==1]['calendarDate'].tolist()
    # 循环一次只得到一个数据，某一个组合某一个月的组合收益
    for i in range(12):
        inf1 =DataAPI.MktEqudAdjGet(secID=x,ticker=u"",tradeDate=cal_list[i],beginDate=u"",\
                                 field=u"secID,ticker,closePrice",pandas="1")
        inf1.columns = ['secID','ticker','closePrice1']
        inf2 =DataAPI.MktEqudAdjGet(secID=x,ticker=u"",tradeDate=cal_list[i+1],beginDate=u"",\
                                 field=u"secID,ticker,closePrice,marketValue",pandas="1")
        inf2.columns = ['secID','ticker','closePrice2','marketValue']
        
        inf = pd.merge(inf1,inf2,on='secID',how ='outer').dropna()
        
        inf['return_month']=inf['closePrice2']/inf['closePrice1']-1  # 获得本月的月度收益率
        
        inf['weight']=inf['marketValue']/(inf['marketValue'].sum()) #以流通市值权重，计算每一只股票的权重收益率
        inf['return_month_weight']=inf['return_month']*inf['weight']
        
        returnMonthly.append(sum(inf['return_month_weight']))    #最后只得到该组合该月份的组合收益率
                
    return returnMonthly
```

```python
#计算108个月，6个组合每一个月份的以市值为权重的组合收益率
all_return = pd.DataFrame(index=range(0,108),columns = ['SH','SM','SL','BH','BM','BL']) 

all_year = [2007,2008,2009,2010,2011,2012,2013,2014,2015]

for i in range(9):
    
    SH,SM,SL,BH,BM,BL=get_6groups(str(int(all_year[i])*10000+531))  #获得每一年度的分组
    
    return_SH = get_returnMonthly(SH,all_year[i])     #return_SH包含本年12个月份的组合收益的list
    return_SM = get_returnMonthly(SM,all_year[i])
    return_SL = get_returnMonthly(SL,all_year[i])
    return_BH = get_returnMonthly(BH,all_year[i])
    return_BM = get_returnMonthly(BM,all_year[i])
    return_BL = get_returnMonthly(BL,all_year[i])
    
    all_return.loc[i*12:i*12+11,'SH']=return_SH  # 将list  储存到 dateframe中的特定行特定列中
    all_return.loc[i*12:i*12+11,'SM']=return_SM
    all_return.loc[i*12:i*12+11,'SL']=return_SL
    all_return.loc[i*12:i*12+11,'BH']=return_BH
    all_return.loc[i*12:i*12+11,'BM']=return_BM
    all_return.loc[i*12:i*12+11,'BL']=return_BL

    all_return['SMB']=(all_return['SH']+all_return['SM']+all_return['SL']-all_return['BH']-all_return['BM']-all_return['BL'])/3
    
    all_return['HML']=(all_return['BH']+all_return['SH']-all_return['BL']-all_return['SL'])/2
    # 循环一次得到某一年12个月的数据
        
all_return        
```

```python
import pandas as pd
import numpy as np

#以中证800指数代表市场利率
RM_Monthly = DataAPI.MktIdxmGet(beginDate=u"20070601",endDate=u"20160531",indexID=u"000906.ZICN",\
                                ticker=u"",field=u"indexID,secShortName,chgPct,endDate",pandas="1")
RM_Monthly = RM_Monthly.sort_values('endDate')

#以银行间同业拆借利率代表市场无风险利率（sirb）
#获取每月末的交易日历,经典的日期转换方式
CAL_list1 = RM_Monthly['endDate'].tolist() 
CAL_list2 = [x.replace('-','') for x in CAL_list1] 

#经典的整合多个dateframe到一个dateframe中的方式
RF_Monthly_list=[]
for i in range(108):
    rf_Monthly = DataAPI.MktIborGet(secID=u"",ticker=u"Shibor1D",tradeDate=CAL_list2[i],\
                                field=u"tradeDate,secID,ticker,rate",pandas="1")
    rf_Monthly['rate']=rf_Monthly['rate']/12  #年收益率必须均转化为月度的收益率，除以12，按照单利计算
    RF_Monthly_list.append(rf_Monthly)
RF_Monthly = pd.concat(RF_Monthly_list,axis = 0)
RF_Monthly = RF_Monthly.sort_values('tradeDate')
#
#获取市场因子RM-RF
RF_Monthly['rate']=RF_Monthly['rate']/100
RM_RF_Monthly = pd.merge(RM_Monthly,RF_Monthly,left_on='endDate',right_on ='tradeDate')
RM_RF_Monthly = RM_RF_Monthly[['endDate','chgPct','rate']]
RM_RF_Monthly.columns = ['endDate','RM_Monthly','RF_Monthly']
RM_RF_Monthly['RM-RF']=RM_RF_Monthly['RM_Monthly']-RM_RF_Monthly['RF_Monthly']

#将三个因子整合到一起，加上日期
RM_RF_Monthly['SMB'] = all_return['SMB']
RM_RF_Monthly['HML'] = all_return['HML']
all_factor = RM_RF_Monthly[['endDate','RM-RF','SMB','HML']]
all_factor
```

```python
# 定义每年5月最后一个交易日的分组函数，分为25组，目的是为得到每个组合的股票
def get_25groups(breakpoint):            #breakpoint为str
    
    #获取每5月最后一个交易日期
    cal_date = DataAPI.TradeCalGet(exchangeCD=u"XSHG",beginDate=str(int(breakpoint)-20),endDate=breakpoint,\
                                    field=u"calendarDate,isOpen",pandas="1")
    cal_date = cal_date[cal_date['isOpen']==1]
    breakpoint = cal_date['calendarDate'].tolist()[-1]   #list的最后一个数值
    
    universe = set_universe('000906.ZICN',date=breakpoint)  #选取universe时候必须注明时间,取当时的中证800
    #每年531日，重新取中证800的成分股
    #对每5月最后一个交易日期的中证800按照市值分为S B 两组
    
    #获取所有股票的市值
    ME= DataAPI.MktEqudGet(secID=universe,ticker=u"",tradeDate=breakpoint,\
                            field=u"secID,ticker,marketValue",pandas="1").dropna()
    
    #获取所有股票的账面市值比
    BP= DataAPI.MktStockFactorsOneDayGet(secID=universe,ticker=u"",tradeDate=breakpoint,\
                            field=u"secID,ticker,PB",pandas="1").dropna()
    
    BP=BP[BP['PB']>0]
    BP['BP']=1/BP['PB']
    BP=BP.drop('PB',axis = 1)
    
    Portifolio_1=[]       #用于存储25个组合的每个组合股票代码的列表，列表组成的列表
    
    Percentile = [0,20,40,60,80,100]
    for i in range(5):
        for j in range(5):
            
            ME_I = ME[(ME['marketValue']>=(np.percentile(ME['marketValue'],Percentile[i])))&\
                      (ME['marketValue']<(np.percentile(ME['marketValue'],Percentile[i+1])))]['secID'].tolist()
            
            BP_I = BP[(BP['BP']>=(np.percentile(BP['BP'],Percentile[j])))&\
                      (BP['BP']<(np.percentile(BP['BP'],Percentile[j+1])))]['secID'].tolist()
            
            ME_BP = set(ME_I).intersection(set(BP_I))
            
            ME_BP=list(ME_BP)
            Portifolio_1.append(ME_BP)  #不可以使用Portifolio_1=Portifolio_1.append(ME_BP),返回25个股票组合的代码列表
            
    return Portifolio_1
```

```python
#计算108个月，每一年12个月 25个组合的组合收益率，组成一个DataFrame  index按照月份顺序，columns按照组合顺序

Portifolio_return = pd.DataFrame(index=range(0,108),columns = range(0,25)) 

all_year = [2007,2008,2009,2010,2011,2012,2013,2014,2015]  # 9年 20070601-20160531  25组合

for i in range(9):   
    for j in range(25):  #必须多使用循环，有时候太多，不可以穷列法
        
        all_Portifolio = get_25groups(str(int(all_year[i])*10000+531))
        a = all_Portifolio[j]
        year_return = get_returnMonthly(a,all_year[i])   # list的功能很强大
        
        Portifolio_return.loc[i*12:i*12+11,int(j)]= year_return  # 将list  储存到 dateframe中的特定行特定列中
        
Portifolio_return        
```

```python
# 25组的组合收益率，每一组减去无风险收益率
RF_Monthly_list = list(RF_Monthly['rate'])  # 转化为list，为下面做准备
net_Portifolio_return = Portifolio_return.sub(RF_Monthly_list,axis=0)  # DataFrame 与list之间的四则运算
net_Portifolio_return
```

检验CAPM模型： Rm-Rf = a + b*(RM-RF)+ e
现在对CAPM模型做回归分析：

```python
import statsmodels.api as sm
#系数
params_a1 = np.zeros(25)
params_b1 = np.zeros(25)
#P值
Pvalues_a1 = np.zeros(25)
Pvalues_b1 = np.zeros(25)
#T值
Tvalues_a1 = np.zeros(25)
Tvalues_b1 = np.zeros(25)
#R2
R2_1=np.zeros(25)
#残差
resid_e1 =np.zeros((25,108))

for i in range(25):
    X =np.zeros((1,108))
    X[0]=all_factor['RM-RF'].tolist()
    
    X = X.T
    X = sm.add_constant(X)

    y = net_Portifolio_return[i].tolist() 
    
    modle = sm.OLS(y,X)  #y,X均为list
    results = modle.fit()
    
    params_a1[i] = results.params[0]
    params_b1[i] = results.params[1]
    
    #P值
    Pvalues_a1[i] = results.pvalues[0]
    Pvalues_b1[i] = results.pvalues[1]
    
    #T值
    Tvalues_a1[i] = results.tvalues[0]
    Tvalues_b1[i] = results.tvalues[1]
      
    #R2
    R2_1[i]=results.rsquared
    
    #残差
    resid_e1[i] =results.resid
```

```python
import statsmodels.api as sm
#系数
params_a2 = np.zeros(25)

params_s2 = np.zeros(25)
params_h2 = np.zeros(25)
#P值
Pvalues_a2 = np.zeros(25)

Pvalues_s2 = np.zeros(25)
Pvalues_h2 = np.zeros(25)
#T值
Tvalues_a2 = np.zeros(25)

Tvalues_s2 = np.zeros(25)
Tvalues_h2 = np.zeros(25)
#R2
R2_2=np.zeros(25)
#残差
resid_e2 =np.zeros((25,108))

for i in range(25):
    X =np.zeros((2,108))
    
    X[0]=all_factor['SMB'].tolist()
    X[1]=all_factor['HML'].tolist()
    X = X.T
    X = sm.add_constant(X)

    y = net_Portifolio_return[i].tolist() 
    
    modle = sm.OLS(y,X)
    results = modle.fit()
    
    params_a2[i] = results.params[0]
   
    params_s2[i] = results.params[1]
    params_h2[i] = results.params[2]
    #P值
    Pvalues_a2[i] = results.pvalues[0]
    
    Pvalues_s2[i] = results.pvalues[1]
    Pvalues_h2[i] = results.pvalues[2]
    #T值
    Tvalues_a2[i] = results.tvalues[0]
   
    Tvalues_s2[i] = results.tvalues[1]
    Tvalues_h2[i] = results.tvalues[2]    
    #R2
    R2_2[i]=results.rsquared
    #残差
    resid_e2[i] =results.resid
```
