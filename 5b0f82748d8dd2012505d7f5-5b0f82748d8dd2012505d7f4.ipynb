```python
#来自量化学堂，第3课 财务分析入门基础课
import pandas as pd

def load_fiscal_data(ticker='603288', end_date='2017-09-30'):
    #提取计算FCFF的各项数据
    """
    1)资产负债表, FdmtBSGet
    TCA: 流动资产合计
    cashCEquiv: 货币资金
    TCL: 流动负债合计
    STBorr: 短期借款
    2)利润表, FdmtISGet
    NIncome: 净利润
    finanExp: 财务费用
    3)现金流量表, FdmtCFGet
    COutfFrInvestA, 投资活动现金流出小计
    CPaidInvest, 投资支付现金
    4)现金流量表附表, FdmtCfsGet
    FAOGPBDepr, 固定资产折旧
    intanAssetsAmor, 无形资产摊销
    """
    #1、DataAPI.FdmtBSGet合并资产负债表 (Point in time)
    bs = DataAPI.FdmtBSGet(ticker=ticker,secID=u"",reportType=u"",endDate=u"",beginDate=u"2013-09-30",\
                                    publishDateEnd=u"",publishDateBegin=u"",\
                                    endDateRep="",beginDateRep="",beginYear="",endYear="",fiscalPeriod="",\
                                    field=u"ticker,endDate,publishDate,fiscalPeriod,TCA,cashCEquiv,TCL,STBorr",pandas="1")
    #从资产负债表里面获取对应股票的流动资产合计、货币资金、流动负债合计、短期借款数据
    bs = bs.sort_values(by=['endDate', 'publishDate', 'fiscalPeriod'], ascending=[True, True, False]).drop_duplicates(subset=['endDate'], keep='first')
    #sort_values选出的数据表格进行排序，依次按关键字'endDate', 'publishDate', 'fiscalPeriod'进行升序、升序、降序排列
    #drop_duplicates按'endDate'关键字进行去重，保留第一次出现的
    # print '资产负债表:','\n',bs.head()
    # print '='*80
    
    #2、DataAPI.FdmtISGet合并利润表 (Point in time)
    income = DataAPI.FdmtISGet(ticker=ticker,secID=u"",reportType=u"",endDate=u"",beginDate=u"2013-09-30", \
                                    publishDateEnd=u"",publishDateBegin=u"",\
                                    endDateRep="",beginDateRep="",beginYear="",endYear="",fiscalPeriod="", \
                                    field=u"ticker,secShortName,endDate,publishDate,fiscalPeriod,NIncome,finanExp",pandas="1")
    #从利润表里面选出对应股票的NIncome: 净利润、finanExp: 财务费用
    income = income.sort_values(by=['endDate', 'publishDate', 'fiscalPeriod'], ascending=[True, True, False]).drop_duplicates(subset=['endDate'], keep='first')
    #与资产负债表相同的排序方式，相同的去重方式
    # print '利润表:','\n',income.head()
    # print '='*80  
    
    #3、合并现金流量表 (Point in time)
    cf = DataAPI.FdmtCFGet(ticker=ticker,secID=u"",reportType=u"",endDate=u"",beginDate=u"2013-09-30", \
                                    publishDateEnd=u"",publishDateBegin=u"",\
                                    endDateRep="",beginDateRep="",beginYear="",endYear="",fiscalPeriod="", \
                                    field=u"ticker,endDate,publishDate,fiscalPeriod,COutfFrInvestA,CPaidInvest",pandas="1")
    #从现金流量表获取对应股票的 COutfFrInvestA, 投资活动现金流出小计、CPaidInvest, 投资支付现金
    cf = cf.sort_values(by=['endDate', 'publishDate', 'fiscalPeriod'], ascending=[True, True, False]).drop_duplicates(subset=['endDate'], keep='first')
    #与资产负债表相同的排序方式，相同的去重方式
    # print '现金流量表:','\n',cf.head()
    # print '='*80 
    
    #4、现金流量表补充表 (Point in time)
    cfs = DataAPI.FdmtCfsGet(ticker=ticker,secID=u"",reportType=u"",endDate=u"",beginDate=u"2013-09-30", \
                            beginYear="",endYear="",fiscalPeriod="", \
                            field=u"ticker,endDate,publishDate,fiscalPeriod,FAOGPBDepr,intanAssetsAmor",pandas="1")
    #从现金流量表补充表获取对应股票的 FAOGPBDepr, 固定资产折旧、intanAssetsAmor, 无形资产摊销
    cfs = cfs.sort_values(by=['endDate', 'publishDate', 'fiscalPeriod'], ascending=[True, True, False]).drop_duplicates(subset=['endDate'], keep='first')
    #与资产负债表相同的排序方式，相同的去重方式
    cfs['fiscalPeriod'] = cfs['fiscalPeriod'].astype(int)
    #fiscalPeriod转化为整形，后面要做判断使用
    cfs['publishDate'] = cfs['publishDate'].apply(lambda x: x[:10])
    #去掉publishDate里面的时间，只保留日期
    # print '现金流量表补充表:','\n',cfs.head()
    # print '='*80 
    
    #5、合并以上四表到一个dataframe，方便后续的计算
    bs = bs.merge(income, left_on=['ticker', 'endDate', 'publishDate', 'fiscalPeriod'], right_on=['ticker', 'endDate', 'publishDate', 'fiscalPeriod'])
    bs = bs.merge(cf, left_on=['ticker', 'endDate', 'publishDate', 'fiscalPeriod'], right_on=['ticker', 'endDate', 'publishDate', 'fiscalPeriod'])
    bs['fiscalPeriod'] = bs['fiscalPeriod'].astype(int)
    bs = bs.merge(cfs, left_on=['ticker', 'endDate', 'publishDate', 'fiscalPeriod'], right_on=['ticker', 'endDate', 'publishDate', 'fiscalPeriod'], how='inner')
    bs = bs.fillna(0.0)
    #缺失值0填充，后面需要对数据进行运算
    print '合并后的dataFrame:','\n',bs.tail()
    print '='*80 

    return bs

def comp_fcff(df, wacc=[0.08,0.09,0.1], gs=[0.0, 0.13, 0.12, 0.11, 0.10, 0.09, 0.08, 0.07, 0.07, 0.07, 0.07, 0.06]):
    #设定平均资本成本和每年增长速度  默认参数
    """
    1) 计算fcff
    2) 计算各期fcff折现，并考虑永续增长
    """
    """
    1)资产负债表, FdmtBSGet
    TCA: 流动资产合计
    cashCEquiv: 货币资金
    TCL: 流动负债合计
    STBorr: 短期借款
    2)利润表, FdmtISGet
    NIncome: 净利润
    finanExp: 财务费用
    3)现金流量表, FdmtCFGet
    COutfFrInvestA, 投资活动现金流出小计
    CPaidInvest, 投资支付现金
    4)现金流量表附表, FdmtCfsGet
    FAOGPBDepr, 固定资产折旧
    intanAssetsAmor, 无形资产摊销
    """
    # 公司自由现金流FCF=净利润NIncome+财务费用finanExp+折旧及摊销（FAOGPBDepr、intanAssetsAmor）-资本开支-净运营资本增加额
    #获取年度数据
    df = df[df['fiscalPeriod'] == 12]
    print '年度数据','\n',df.tail()
    print '='*80 
    #用最新一期的年度数据来计算FCFF
    fcff0 = df['NIncome'].iloc[-1] + df['finanExp'].iloc[-1] + df['FAOGPBDepr'].iloc[-1] + df['intanAssetsAmor'].iloc[-1]
    #资本开支的计算
    cap_exp = -(df['COutfFrInvestA'].iloc[-1] - df['CPaidInvest'].iloc[-1])
     #净运营资本增加额的计算
    nwc_chg = (df['TCA'] - df['cashCEquiv'] - df['TCL'] + df['STBorr']).diff(1).iloc[-1]
    #diff(1)函数是用原数据减去原数据在行往下偏1的dataframe，iloc[-1]取出最后一个数据，也就是计算出净增加额
    #df.diff(1)相当于df-df.shift(1)
    fcff0 = fcff0 - cap_exp - nwc_chg
    print('fcff0: {0:.2f} cap_exp:{1:.2f} nwc_chg:{2:.2f}'.format(fcff0, cap_exp, nwc_chg))
    #{:.2f}保留小数点后两位
    columns = ['year', 'growth', 'fcff'] + ['discount0={0}'.format(x) for x in wacc]
    #定义输出数据的列标签，'discount0={0}'.format(x) for x in wacc使用非常好
    data = []   #定义一个空列表
    for year, g in zip(range(len(gs)), gs):
        #zip()函数的定义，从参数中的多个迭代器取元素组合成一个新的迭代器，当a与b的行数或列数不同时，取两者结构中最小的行数和列数，依照最小的行数和列数将对应位置的元素进行组合
        #两个元素一起循环，加快速度
        fcff = fcff0 if year == 0 else fcff
        fcff = fcff * (1 + g)
        # 永续增长
        if year == len(gs)-1:
            data.append(['eternal', g, fcff] + [fcff * (1 + g) / (discount - g) / (1+discount) ** year for discount in wacc])
        else:
            data.append([year, g, fcff] + [fcff / (1+discount)**year for discount in wacc])
            #相当于套用了一个小循环，使用非常好
            #此时data是一个列表里面套用列表
    enp_fcff = pd.DataFrame(data=data, columns=columns)
    for x in wacc:
        key = 'discount0={0}'.format(x)
        print('discount={0} enterprise fcff={1:.2f}E'.format(x, sum(enp_fcff[key])/1e8))
    return enp_fcff

# 可以看出来通过我们假设的企业增长率及折现成本，海天味业的市值在1275亿~2569亿，当前市值在1495亿，较为合理
fiscal_df = load_fiscal_data(ticker='603288')
enp_fcff = comp_fcff(fiscal_df)
enp_fcff
```
