**问题是，如何判断是否是震荡行情？进入趋势行情，则认为震荡行情结束。**
上涨趋势行情有以下特征：
1、价位创出新高，高点更高，低点也更高。
2、价位长期（大于15天、1个月甚至更久）站上80、90百分位价格。
3、如有交易量的放大，则更加确认上涨趋势行情。
