# 2850底是被外资砸出来的？——沪股通数据简单分析

今天是沪港通成立一周年，这一年里，北上的沪股通投资者一直被当做「聪明的」投资者，碰巧发现通联可以购买沪港通的数据，并且是免费的（见「恒生聚源·港股」），我就简单跑了跑结果（主要估算了沪股通外资的持股份额以及持股成本线）与大家分享，有想法的人可以以它为指标或者反向指标？

```python
#####importing package#######
import numpy as np
import pandas as pd
import datetime as dt
from datetime import timedelta
from matplotlib import pyplot as plt
from CAL.PyCAL import *
```

```python
today = Date.todaysDate()
endDate = today.toDateTime().strftime('%Y%m%d')   
```

```python
#得到每日沪股通的流入流出金额
def dailyVolumn(date):
    dailyQuota = 13000000000 #the quota is fixed
    dailyBalance = DataAPI.JY.HKSHKQuotaJYGet(endDate=date,tradingTypeCd=u"1",field=u"dailiBalance",pandas="1")
    dailyVolumn = np.array(dailyQuota - dailyBalance['dailiBalance'])
    return dailyVolumn
```

```python
closeIndex = DataAPI.MktIdxdGet(ticker='000001',beginDate='20141117',endDate=endDate,field=["tradeDate","closeIndex"],pandas="1")
data = pd.DataFrame({'tradeDate':closeIndex['tradeDate'], 'indexPrice':closeIndex["closeIndex"], 'dailyVolumn':np.array([0] * len(closeIndex)),'accumVolumn':np.array([0] * len(closeIndex)),'dailyShare':np.array([0] * len(closeIndex)),'accumShare':np.array([0] * len(closeIndex)),'meanPrice':np.array([0] * len(closeIndex))})
#data.index = closeIndex['tradeDate']
for i in range(len(data)):
    date = data.loc[i,'tradeDate']
    try:
        dV = dailyVolumn(date.replace('-',''))
        data.loc[i,'dailyVolumn'] = dV
        if i==0:                
            data.loc[i,'accumVolumn'] = data.loc[i,'dailyVolumn']
            data.loc[i,'dailyShare'] = data.loc[i,'dailyVolumn'] / data.loc[i,'indexPrice']
            data.loc[i,'accumShare'] = data.loc[i,'dailyShare']
            data.loc[i,'meanPrice'] = data.loc[i,'accumVolumn'] / data.loc[i,'accumShare']
        else:                
            data.loc[i,'accumVolumn'] = data.loc[i,'dailyVolumn'] + data.loc[i-1,'accumVolumn']
            data.loc[i,'dailyShare'] = data.loc[i,'dailyVolumn'] / data.loc[i,'indexPrice']
            data.loc[i,'accumShare'] = data.loc[i,'dailyShare'] + data.loc[i-1,'accumShare']
            data.loc[i,'meanPrice'] = data.loc[i,'accumVolumn'] / data.loc[i,'accumShare']
    except:
        data.loc[i,'accumVolumn'] = data.loc[i,'dailyVolumn'] + data.loc[i-1,'accumVolumn']            
        data.loc[i,'dailyShare'] = data.loc[i,'dailyVolumn'] / data.loc[i,'indexPrice']
        data.loc[i,'accumShare'] = data.loc[i,'dailyShare'] + data.loc[i-1,'accumShare']
        data.loc[i,'meanPrice'] = data.loc[i,'accumVolumn'] / data.loc[i,'accumShare']
        continue
data.index = data['tradeDate']        
        
```

```python
data[['indexPrice','meanPrice']].plot(figsize=(14,5),linewidth=2)
plt.figure(2)
data['accumShare'].plot(figsize=(14,3),color='red',title='accumShare',linewidth=2)
```

蓝线是上证指数，绿线是沪股通外资的估算成本线。可以看到，当8月24日指数跌破3000点时，即沪股通外资的成本线时，外资有明显的补仓抄底行为，似乎2850底是被外资砸出来的？

然而，随着指数反弹到3500点时，外资逐渐获利离场。同时也需注意到，在3500到4000点之间的区间，外资没有买入股票的行为，此处被套住的似乎是散户。当指数大于3500点时，外资似乎没有意愿继续购入股票？否则是给散户解套？
