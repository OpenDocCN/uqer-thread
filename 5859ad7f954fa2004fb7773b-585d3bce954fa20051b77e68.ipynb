### 处理数据

```python
def max_arr(arr):
    try:
        se=talib.MAX(np.array(arr),30)
    except:
        se=np.array([np.nan]*len(arr))
    return se    
def min_arr(arr):
    try:
        se=talib.MIN(np.array(arr),30)
    except:
        se=np.array([np.nan]*len(arr))
    return se
fields = ['tradeDate','closeIndex', 'highestIndex','lowestIndex', 'turnoverVol','CHG','CHGPct']
stock = '000300'
#tradeDate是交易日、closeIndex是收盘指数、highestIndex是当日最大指数，lowestIndex是当日最小指数，CHG是涨跌
index_raw = DataAPI.MktIdxdGet(ticker=stock,beginDate=u"2006-03-01",endDate=u"2014-12-01",field=fields,pandas="1")
index_date = index_raw.set_index('tradeDate')
index_date = index_date.dropna()
index_date['max_difference'] = index_date['highestIndex'] - index_date['lowestIndex']
index_date['max_of_30day']=index_date.apply(max_arr)['highestIndex']
index_date['min_of_30day']=index_date.apply(min_arr)['lowestIndex']
index_date['max_difference_of_30day']=index_date.apply(max_arr)['max_difference']
index_date['closeIndex_after30days']=np.nan
index_date['closeIndex_after30days'][0:-30]=np.array(index_date['closeIndex'][30:])
index_date = index_date.dropna()   #去掉前30个和后30个无效的数据。
# lables_raw = index_date['closeIndex_after30days'] #提取出需要预测的数据
# lables = (index_date['closeIndex_after30days'] > index_date['closeIndex']*1.1) #为分类处理数据，判断30天后的收盘价是否大于今日收盘价
# index_date['closeIndex_after30days'] = index_date['closeIndex_after30days'] - index_date['closeIndex']
index_date[(index_date['closeIndex_after30days']<index_date['closeIndex']*1.1)&
          (index_date['closeIndex_after30days']>index_date['closeIndex']*0.9)]=np.nan
index_date = index_date.dropna() 
index_date['closeIndex_after30days'][index_date['closeIndex_after30days']>index_date['closeIndex']*1.1] = 'up'
index_date['closeIndex_after30days'][index_date['closeIndex_after30days']<index_date['closeIndex']*0.9] = 'down'
index_date = index_date.dropna() 
lables = index_date['closeIndex_after30days']

features = index_date.drop(['closeIndex_after30days'],axis = 1) #在特征值中去掉我们要预测的数据
```

```python
from sklearn import cross_validation
from sklearn import preprocessing
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn import svm

sp = int(len(features)*5./6.)

features_1 = features[:sp]
features_2 = features[sp:]
lables_1 = lables[:sp]
lables_2 = lables[sp:]

scaler1 = preprocessing.StandardScaler().fit(features_1)
features_scaler_1 = scaler1.transform(features_1)
scaler2 = preprocessing.StandardScaler().fit(features_2)
features_scaler_2 = scaler2.transform(features_2)
#上面4行代码用来标准化数据

X_train,X_test, y_train, y_test = cross_validation.train_test_split(features_scaler_1, lables_1, test_size = 0.2, random_state = 0)
```

### 在未调参之前，我们先获取一次准确率：

```python
clf_svm = svm.SVC()   #使用SVM分类器来判断涨跌
clf_svm.fit(X_train, y_train)
print "样本内预测准确率为：%0.2f" % (clf_svm.score(X_test, y_test))
print "样本外预测准确率为：%0.2f" % (clf_svm.score(features_scaler_2, lables_2))
```

### 然后调C值，这里我是先让C在1~200的range跑:发现对样本内数据表现良好，但是对样本外数据表现就很差了。

```python
i_list = []
score_list1 = []
score_list2 = []
for i in xrange(1,200,1):
    i=i/10.
    clf_svm = svm.SVC(C = i)   #使用SVM分类器来判断涨跌
    clf_svm.fit(X_train, y_train)
    i_list.append(i)
    score_list1.append(clf_svm.score(X_test, y_test))
    score_list2.append(clf_svm.score(features_scaler_2 , lables_2))

score_list_df =  pd.DataFrame({'C_score':i_list,'in_sets':score_list1,'out_of_sets':score_list2})
score_list_df.plot(x='C_score',title='accuracy change with c')
```

### 然后是gamma值，和C值调参上也是同理。

```python
i_list = []
score_list1 = []
score_list2 = []
for i in range(1,100,1):
    i=i/10.
    clf_svm = svm.SVC(C=1 ,  gamma = i)  
    clf_svm.fit(X_train, y_train)
    i_list.append(i)
    score_list1.append(clf_svm.score(X_test, y_test))
    score_list2.append(clf_svm.score(features_scaler_2, lables_2))

score_list_df =  pd.DataFrame({'gamma_score':i_list,'in_sets':score_list1,'out_of_sets':score_list2})
score_list_df.plot(x='gamma_score',title='accuracy change with gamma')
```

### 虽说没什么卵用……还是假吧意思的比对一下不同核函数下的准确率吧。结果！这次不是高斯核表现最好了！

```python
i_list = []
score_list1 = []
score_list2 = []

kernels  =  ['linear', 'rbf','sigmoid']
for i in kernels :
    clf_svm = svm.SVC( kernel = i )   
    clf_svm.fit(X_train, y_train)
    i_list.append(i)

    score_list1.append(clf_svm.score(X_test, y_test))
    score_list2.append(clf_svm.score(features_scaler_2, lables_2))

score_list_df =  pd.DataFrame({'kernels':i_list,'in_sets':score_list1,'out_of_sets':score_list2})
score_list_df.plot(x='kernels',title='accuracy change with kernels',kind='bar')
```

### 然后在最优解的基础上再次计算一次准确率

```python
clf_svm = svm.SVC(C=4,gamma=0.05,kernel='rbf')  
clf_svm.fit(X_train, y_train)
print "样本内预测准确率为：%0.2f" % (clf_svm.score(X_test, y_test))
print "样本外预测准确率为：%0.2f" % (clf_svm.score(features_scaler_2, lables_2))
```

### 改变时间周期，观察准确率在样本内测试集和样本外测试集上的表现。

发现随着时间的变化，模型的准确率越来越低，大约在3个月以后准确率低于阈值（0.62）又似乎呈周期性变化？不过我们同时也知道了另一件事，如果不及时更新数据，预测大盘的能力会越来越差。

```python
i_list = []
score_list1 = []
score_list2 = []
for i in xrange(200):
    num_now = len(features) + i -300
    
    features_1 = features[:num_now]
    features_2 = features[num_now:num_now+100]
    lables_1 = lables[:num_now]
    lables_2 = lables[num_now:num_now+100]

    scaler1 = preprocessing.StandardScaler().fit(features_1)
    features_scaler_1 = scaler1.transform(features_1)
    scaler2 = preprocessing.StandardScaler().fit(features_2)
    features_scaler_2 = scaler2.transform(features_2)

    X_train,X_test, y_train, y_test = cross_validation.train_test_split(features_scaler_1, lables_1, test_size = 0.2, random_state = 0)
    
    clf_svm = svm.SVC(C=4,gamma=0.05,kernel='rbf')   #使用SVM分类器来判断涨跌
    
    clf_svm.fit(X_train, y_train)
    i_list.append(features_2.index[0])
    score_list1.append(clf_svm.score(X_test, y_test))
    score_list2.append(clf_svm.score(features_scaler_2, lables_2))

score_list_df =  pd.DataFrame({'date':i_list,'in_sets':score_list1,'out_of_sets':score_list2})
score_list_df.plot(x='date',title='Accuracy with sets')
```
