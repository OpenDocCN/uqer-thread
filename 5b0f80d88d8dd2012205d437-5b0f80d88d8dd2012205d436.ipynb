# 第2课  带约束条件最优解

```python
#来自量化学堂 第2课 Python科学计算工具包

'''
作业一：定义函数 f(x,y)=2xy+2x−x2−2y2
带有约束条件：

2x−y=0
y−1⩾0
求解f(x,y)的最大值
'''
#思路：
#1、定义函数
#2、求雅克比矩阵
#3、定义约束条件
#4、求解最优解
#5、图像展示

import numpy as np
from scipy import optimize
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter

#1、定义函数
def f(x,sign=1):
    return sign * (2 * x[0] * x[1] + 2 * x[0] - x[0] ** 2 - 2 * x[1] ** 2)

#2、求雅克比矩阵
def j(x, sign=1.0):
    x0,x1 = x.tolist()
    return [sign*(2*x[1] + 2 - 2*x[0]),sign*(2*x[0] - 4*x[1])]
                                            
#3、定义约束条件
cons = (
    {'type': 'eq', 'fun': lambda x: 2 * x[0] - x[1]},
    {'type': 'ineq', 'fun': lambda x: x[1] - 1}
)                   

#4、求解最优解
x0 = np.array([0, 0])
result_wi_cons = optimize.minimize(f, x0, args=(-1.0,), jac=j, method='SLSQP',constraints=cons).x
print "x: %s, y: %s" % (result_wi_cons[0], result_wi_cons[1])   
```

```python
#5、图像展示
#函数与约束条件的展示

#a、创建Axes 3D对象
fig = plt.figure()
ax = fig.gca(projection='3d')

#b、坐标范围设定
x = np.arange(-5, 5, 0.1)
y = np.arange(-5, 5, 0.1)
x, y = np.meshgrid(x, y)
z = f([x, y])

#c、进行图像绘制
#函数图像绘制
surf = ax.plot_surface(x, y, z, cmap=cm.coolwarm, linewidth=0., antialiased=True)
# 绘制约束条件
ax.plot(np.arange(-2, 2, 0.1), 2 * np.arange(-2, 2, 0.1)) # 条件1
x_1 = np.arange(-5, 5, 0.1)
y_1 = 1
z_1 = np.arange(-140, 20, 1)
x_1, z_1 = np.meshgrid(x_1, z_1)
ax.plot_surface(x_1, 1, z_1, ) # 条件2
#绘制最优解
ax.scatter(result_wi_cons[0], result_wi_cons[1], f(result_wi_cons), c='y')

#d、设置图形展示效果
fig.colorbar(surf, shrink=1, aspect=5)
ax.view_init(elev=45, azim=45) # 改变绘制图像的视角,即相机的位置,azim表示转动z轴，elev表示转动y轴
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

#e、显示出来
fig.show()
```
