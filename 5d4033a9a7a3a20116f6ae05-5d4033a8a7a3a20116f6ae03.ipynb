## 策略说明
### Beta简介
市场指数通常都是市值加权，如果采用非市值加权，得到的贝塔就是Smart Beta，以下分别从等权组合，最小方差组合，风险平价组合，最大多元组合这4个角度进行权重优化
### 原理与算法
#### 1 等权重
$$w_i=\frac{1}{N}$$
#### 2 最小方差
用于保证整体方差风险最小，用以下优化问题表示：
$$w^*=min(\frac{1}{2}\omega'\Omega\omega)$$
$$s.t. w'1=1$$
$$\omega_i \geq 0$$
$\Omega$ 为各资产协方差举证，$\boldsymbol{\omega}$为各资产权重列向量
#### 3 风险平价
保证每个资产对组合整体风险的贡献都一样，用如下优化问题表示：
$$min(f(\omega))=\sum_{i=1}^{N}\sum_{j=1}^{N}[\omega_i(\Omega\omega)_i-\omega_j(\Omega\omega)_j]$$
$$s.t. w'1=1$$
$$\omega_i\geq0$$
#### 4 最大多元化
首先假设组合投资没有分散风险的功能，那么组合波动率就可以像像收益率一样加权求得，但在实际情况下组合投资可以分散风险，那么最大多元化（也就是最大分散度）可以被定义为加权波动率与真实波动率的比值，用以下优化过程表示：
$$max D(\omega)=\frac{\omega'\sigma}{\sqrt{\omega'\Omega\omega}}$$
$$s.t. w'1=1$$
$$\omega_i\geq0$$
$\sigma$为各资产标准差列向量

```python
# 协防差函数
def get_covmat(tickers, date, periods):
    '''
    输入ticker+日期+过去天数，获得以此计算出来的年化协方差矩阵
    '''
    start_date = shift_date(date, periods) 
    return_mat = DataAPI.MktEqudAdjGet(ticker=tickers, beginDate=start_date, endDate=date, field=['ticker', 'tradeDate', 'closePrice'], pandas='1')
    return_mat = return_mat.pivot(index='tradeDate', columns='ticker', values='closePrice').pct_change().fillna(0.0)
    return return_mat.cov()*250 
```

```python
# 定义输入协方差矩阵，得到不同优化方法的权重配置
def get_smart_weight(cov_mat, method='min variance', wts_adjusted=False):
    '''
    功能：输入协方差矩阵，得到不同优化方法的权重
    Args:
    cov_mat: 协方差矩阵。DataFrame，index和columns均为资产名称
    method: 优化方法，可选的有min variance, risk parity, max diversification, equal weight
    Return:
    pd.Series index为资产名， value为weight
    PS:
    依赖于scipy package
    '''
    if not isinstance(cov_mat, pd.DataFrame):
        raise ValueError('cov_mat should be DataFrame')
        
    omega = np.matrix(cov_mat.values)  # 协方差矩阵
    # 定义目标函数
    def fun1(x):
        '''
        最小方差法
        最小值优化
        '''
        return np.matrix(x) * omega * np.matrix(x).T  # [1xn] * [nxn] * [n*1] = [1x1], x为[1xn]
    
    def fun2(x):
        '''
        风险平价
        最小值优化
        '''
        tmp = (omega * np.matrix(x).T).A1  
        risk = x * tmp  # [nx1]
        delta_risk = [sum((i-risk) ** 2) for i in risk]
        return sum(delta_risk)
        
    def fun3(x):
        '''
        最大多元化
        最大值优化
        '''
        den = x * omega.diagonal().T  
        num = np.sqrt(np.matrix(x) * omega * np.matrix(x).T)  
        return num/den  
    
    # 初始值 + 约束条件
    x0 = np.ones(omega.shape[0]) / omega.shape[0]
    bnds = tuple((0, None) for x in x0)  
    cons = ({'type': 'eq', 'fun': lambda x: sum(x)-1})  
    options = {'disp': False, 'maxiter': 1000, 'ftol': 1e-20}  

    if method == 'min variance':
        res = minimize(fun1, x0, bounds=bnds, constraints=cons, options=options, method='SlSQP')
    elif method == 'risk parity':
        res = minimize(fun2, x0, bounds=bnds, constraints=cons, options=options, method='SlSQP')
    elif method == 'max diversification':
        res = minimize(fun3, x0, bounds=bnds, constraints=cons, options=options, method='SlSQP')
    elif method == 'equal weight':
        return pd.Series(index=cov_mat.index, data=1.0/cov_mat.shape[0])
    else:
        raise ValueError('method has not be specified')

    # 权重调整
    if res['success'] == False: 
        print(res['message'])
    wts = pd.Series(index=cov_mat.index, data=res['x'])
    
    if wts_adjusted == True:
        wts = wts[wts >= 0.0001]
        return wts / wts.sum() * 1.0
    elif wts_adjusted == False:
        return wts
    else:
        raise ValueError('wts_adjusted should be True/False')
```

```python
# 定义获取成分股的函数如下
def get_idx_cons(idx, date):
    '''
    功能:
        获取指数某一天的成分股列表
    Args:
        idx: 指数，'xxxxxx'型。可以是list，传入多个指数返回的是并集，不需要考虑重复
        date: 日期，'yyyymmdd'
    Return:
        list of tickers
    '''
    try:
        data = DataAPI.IdxConsGet(ticker=idx, intoDate=date, field='consTickerSymbol')['consTickerSymbol']
    except:
        raise ValueError('some error happend')
    
    return list(set(data))
```

```python
# 定义获取调仓日的函数
def get_dates(start_date, end_date, frequency='daily'):
    '''
    function:
        输入起始日期和频率，得到日期列表
    Args:
        start_date: 开始日期
        end_date: 截止日期
        frequency: 频率，daily为所有交易日，daily1为所有自然日，weekly为每周最后一个交易日，weekly2为每隔两周，monthly为每月最后一个交易日，quarterly为每季最后一个交易日
    Return:
        日期列表list
    '''
    data = DataAPI.TradeCalGet(exchangeCD='XSHG', beginDate=start_date, endDate=end_date, field=['calendarDate', 'isOpen', 'isWeekEnd', 'isMonthEnd', 'isQuarterEnd'])
    if frequency == 'daily':
        data = data[data['isOpen'] == 1]
    elif frequency == 'daily1':
        pass
    elif frequency == 'weekly':
        data = data[data['isWeekEnd'] == 1]
    elif frequency == 'weekly2':
        data = data[data['isWeekEnd'] == 1]
        data = data[0: data.shape[0]: 2]
    elif frequency == 'monthly':
        data = data[data['isMonthEnd'] == 1]
    elif frequency == 'quarterly':
        data = data[data['isQuarterEnd'] == 1]
    else:
        raise ValueError('调仓频率错误')
    date_list = data['calendarDate'].values.tolist()
    return date_list
```

```python
# 定义日期前后移动函数
def shift_date(date, n, direction='back'):
    '''
    function:
        给定date，获取该交易日前或后n个交易日对应的交易日
    Args
        date: str 'yyyymmdd'
        n: 移动n个单位
        director: 移动方向，back or forward
    Return:
        对应的交易日
    '''
    last_two_year = str(int(date[:4])-3) + '0101'
    forward_two_year = str(int(date[:4])-3) + '1231'
    if direction == 'back':
        date_list = get_dates(last_two_year, date, 'daily')
        return date_list[len(date_list)-1-n]
    if direction == 'forward':
        date_list = get_dates(date, forward_two_year, 'daily')
        return daye_list[n]
    else:
        raise ValueError('direction shoud be back/forward')
```

```python
# 定义获取股票收益率的函数
def get_ticker_period_rtn(tickers, start_date, end_date):
    '''
    Function:
        输入tickers和起始日期，获得在此区间的daily return
    Args:
        ticker
        start_date
        end_date
    Return:
        pd.DataFrame
    '''
    begin_date = shift_date(start_date, 1) # 向前推移一天保证第一天也有收益
    data = DataAPI.MktEqudAdjGet(ticker=tickers, beginDate=begin_date, endDate=end_date, field=['ticker', 'tradeDate', 'closePrice'])
    daily_rtn = data.pivot(index='tradeDate', columns='ticker', values='closePrice').pct_change().fillna(0.0)
    return daily_rtn
```

```python
# 定义计算最大回撤的函数
def cal_maxdrawdown(data):
    '''
    Function:
        给定净值数据，返回最大回撤
    Args:
        data: list/np.array/pd.Series/pd.DataFrame，净值曲线，初始值为1
    Return:
        list/np.array/pd.Series返回float
        pd.DataFrame返回pd.DataFrame
    '''
    if isinstance(data, list):
        data = np.array(data)
    if isinstance(data, pd.Series):
        data = data.values
        
    def get_mdd(values):  # values为np.array的的净值曲线，初始金为1
        dd = [values[i:].min() / values[i] - 1 for i in range(len(values))]
        return abs(min(dd))
    
    if not isinstance(data, pd.DataFrame):
        return get_mdd(data)
    else:
        return data.apply(get_mdd)
```

```python
# 定义计算策略评价的各项指标
def cal_indicators(df_daily_return):
    '''
    Function:
        给定daily return，计算各组合的评价指标，包括：年化收益率、年化标准差、夏普率、最大回撤
    Args:
        df_daily_return
    '''
    df_cum_value = (df_daily_return + 1).cumprod()  # 累乘计算总收益
    res = pd.DataFrame(index=['年化收益率', '年化标准差', '夏普值', '最大回撤'], columns=df_daily_return.columns, data=0.0)
    res.loc['年化收益率'] = (df_daily_return.mean() * 250).apply(lambda x: '%.2f%%' % (x*100))  # 算术平均法计算年化收益率
    res.loc['年化标准差'] = (df_daily_return.std() * np.sqrt(250)).apply(lambda x: '%.2f%%' % (x*100))  
    res.loc['夏普值'] = (df_daily_return.mean() / df_daily_return.std() * np.sqrt(250)).apply(lambda x: np.round(x, 2))
    res.loc['最大回撤'] = cal_maxdrawdown(df_cum_value).apply(lambda x: '%.2f%%' % (x*100))
    return res
```

## 历史回测
回测区间从2007年1月1日2018年2月13日，选取上证50成分股为标的，调仓频率为每季度末换仓，用过去500个交易日的数据估计协方差矩阵。

```python
start_date = '2007-01-01'
end_date = '2019-07-19'
idx_code = '000016'  # 上证50
cov_Ndays = 500  # 估计协方差矩阵所用的历史天数
# 获取上证50历史净值
idx_data = DataAPI.MktIdxdGet(ticker=idx_code, beginDate=start_date, endDate=end_date, field=['tradeDate', 'closeIndex'])
idx_data['values'] = idx_data['closeIndex'] / idx_data['closeIndex'][0]
idx_data = idx_data.set_index('tradeDate')
result = pd.DataFrame(index=idx_data.index, columns=['SH50', 'equal weight', 'min variance', 'risk parity', 'max diversification'])
result['SH50'] = idx_data['values']
# 回测
date_list = sorted(get_dates(start_date, end_date, 'quarterly') + [start_date, end_date])
for i in range(len(date_list)-1):
    print(date_list[i])
    current_period = date_list[i]
    next_period = date_list[i+1]  # len(date_list)-1 保证不会超出范围
    tickers = get_idx_cons(idx_code, current_period)
    cov_mat = get_covmat(tickers, current_period, cov_Ndays)
    # 权重优化
    for j in ['min variance', 'risk parity', 'max diversification', 'equal weight']:
        wts = get_smart_weight(cov_mat, method=j, wts_adjusted=True)
        daily_rtn = get_ticker_period_rtn(wts.index.tolist(), current_period, next_period)
        daily_rtn.ix[0] = 0.0
        cum_rtn = np.dot((daily_rtn+1).cumprod(), wts)
        if i == 0:
            result.loc[daily_rtn.index[2:], j] = cum_rtn[2:] * 100
        else:
            result.loc[daily_rtn.index, j] = cum_rtn * result.loc[daily_rtn.index[0], j]
# 格式调整+指标计算
indicators = cal_indicators(result[['SH50', 'equal weight', 'min variance', 'risk parity', 'max diversification']].pct_change().dropna())
```
