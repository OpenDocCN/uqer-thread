# 基金定投通用demo，简单实用

#### 应朋友之邀，做一个基金定投的分析小demo，基本足够研究不同基金的定投走势了

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def caic_return(start, end, period, rf_rate, amt, fund_code):
    # 基金份额价值数据
    fund_data = DataAPI.MktFunddAdjBfGet(secID=u"",ticker=fund_code,beginDate=start,endDate=end,field=u"",pandas="1")[['secID','secShortName','tradeDate','closePrice']]
    # 日历数据
    date = DataAPI.TradeCalGet(exchangeCD=u"XSHG,XSHE",beginDate=start,endDate=end,field=u"",pandas="1")
    date = date[['calendarDate','isWeekEnd','isMonthEnd']]
    # 合并两个DataFrame
    merged = fund_data.merge(date, left_on='tradeDate', right_on='calendarDate', how='inner')
    merged = merged[~merged.duplicated()].reset_index(drop=True) # 剔除重复的数据
    # 按周或者月调整无风险收益率
    if period == 'week':
        tmp = merged[merged['isWeekEnd'] == 1]
        rf_rate = rf_rate / 52
    elif period == 'month':
        tmp = merged[merged['isMonthEnd'] == 1]
        rf_rate = rf_rate / 12
    tmp['share'] = 0 # 份额
    tmp['total_value'] = 0 # 份额总价值
    tmp['rf_value'] = 0 # 储蓄总价值
    tmp.share = amt / tmp.closePrice # 每次定投的价值为：定投金额 / 当日份额净值
    tmp['total_share'] = tmp.share.cumsum() # 计算累积份额
    tmp.total_value = tmp.closePrice * tmp.total_share # 累积份额的价值为：当日份额净值 * 持有总份额
    rf_value = amt # 初始化储蓄价值
    for idx, row in tmp.iterrows():
        tmp.at[idx,'rf_value'] = rf_value
        rf_value = amt + rf_value*(1 + rf_rate) # 每次储蓄的金额 + 储蓄总额*（1+利息） 
    tmp.tradeDate = pd.to_datetime(tmp.tradeDate)
    tmp = tmp.set_index('tradeDate')
    final_total_value = tmp.total_value[-1]
    return tmp, final_total_value
```

#### 改变基金的对应参数，就可以看定投结果

```python
start = "20140521" # 数据开始日期
end = "20190902" # 数据结束日期
period = 'month' # 定投周期，week 或 month
rf_rate = 0.03 # 无风险收益率
fund_code = '163402' # 基金代码，这个就找自己关注的就好
amt = 2000 # 定投金额
result, final_total_value = caic_return(start, end, period, rf_rate, amt, fund_code)
```

```python
result[['total_value','rf_value']].plot()
plt.title('Fund V.S. Saving')
fina_rf_value = result.rf_value[-1]
total_saving = len(result) * amt
print('定投最终受益为：{}元，储蓄最终受益为：{}元，总投入金额为：{}元'.format(fina_total_value, fina_rf_value, total_saving))
```

#### 还有一种比较无脑的办法，可以拿所有有数据的基金，比较哪一个基金定投的历史收益最高。
#### 优矿的接口目前数据有800个左右，足够分析和投资用了。

```python
# 取基金的代码列表
fund_list = DataAPI.MktFunddAdjBfGet(secID=u"",ticker="",beginDate=end,endDate=end,field=u"",pandas="1")['ticker'].tolist()
```

```python
start = "20130101" # 数据开始日期 P.S. 数据最好选取包含牛熊市的情况，这样能更全面的分析基金的表现
end = "20190902" # 数据结束日期
period = 'month' # 定投周期，week 或 month
rf_rate = 0.03 # 无风险收益率
amt = 2000 # 定投金额
fund_ret = 0 # 记录最大的基金收益
for i in fund_list[:10]: # 这里我只取了前10个基金，有兴趣的话可以多跑一些
    fund_code = i 
    result, final_total_value = caic_return(start, end, period, rf_rate, amt, fund_code)
    # 以最终收益为参考，取收益最大的基金为best_one
    if final_total_value > fund_ret:
        fund_ret = final_total_value
        best_one = result
        
best_one[['total_value','rf_value']].plot()
plt.title('Fund V.S. Saving')
fina_rf_value = best_one.rf_value[-1]
total_saving = len(best_one) * amt
print('定投最终受益为：{}元，储蓄最终受益为：{}元，总投入金额为：{}元'.format(fund_ret, fina_rf_value, total_saving))
```
