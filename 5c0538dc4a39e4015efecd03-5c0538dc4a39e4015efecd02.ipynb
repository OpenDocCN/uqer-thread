```python
from pandas import Series, DataFrame
import datetime
import time
from scipy.stats import norm
from math import sqrt, pi,log, e
from CAL.PyCAL import *


def option_delta(data):
    vol_data=data 
    #vol_data = data[data.ticker == ID][data.tradeDate == t]
    r = 0.03
    q = 0.01
    types = vol_data.contractType.values[0]
    if types == 'CO' :
        opt_types = 1
    else :
        opt_types = -1
    strike = vol_data.strikePrice.values[0]
    spot = vol_data.varpreClosePrice.values[0]
    
    exp_date_str = vol_data.expDate.values[0]     # 期权行权日期
    eval_date_str = vol_data.tradeDate.values[0]   # 期权交易日期
    mat = Date.parseISO(exp_date_str) 
    evd = Date.parseISO(eval_date_str) 
    time_to_maturity = float(mat - evd + 1.0)/365.0 
    
    settle = vol_data.preClosePrice.values[0]
    
    calculated_vol = BSMImpliedVolatity(opt_types, strike, spot, 0.03, 0.01, time_to_maturity, settle)      # 计算隐含波动率
    # 返回的是dataframe格式，要转化为数值型
    calculated_vol = calculated_vol.iloc[0,0]
    # 期权的时间价值为负时(此种情况在50ETF期权里时有发生)，没法通过BSM模型计算隐含波动率，故此时将期权隐含波动率设为0.0，实际上，此时的隐含波动率和各风险指标并无实际参考价值
     # 但是为了防止分母为0，故设置为0.01%
    if calculated_vol is None:
        calculated_vol = 0.0001
    
    vol=calculated_vol
    T = time_to_maturity
    d1 = (log(spot/strike) + (0.03 - q + 0.5 * (vol**2)) * T)/(vol * (T**0.5))

    dfq = e**(-q*T)
    if opt_types == 1:
        return(dfq * norm.cdf(d1))
    else:
        return(dfq*(norm.cdf(d1)-1))
    
    
def option_iv(data):
    vol_data=data 
    #vol_data = data[data.ticker == ID][data.tradeDate == t]
    r = 0.03
    q = 0.01
    types = vol_data.contractType.values[0]
    if types == 'CO' :
        opt_types = 1
    else :
        opt_types = -1
    strike = vol_data.strikePrice.values[0]
    spot = vol_data.varpreClosePrice.values[0]
    
    exp_date_str = vol_data.expDate.values[0]     # 期权行权日期
    eval_date_str = vol_data.tradeDate.values[0]   # 期权交易日期
    mat = Date.parseISO(exp_date_str) 
    evd = Date.parseISO(eval_date_str) 
    time_to_maturity = float(mat - evd + 1.0)/365.0 
    
    settle = vol_data.preClosePrice.values[0]
    
    calculated_vol = BSMImpliedVolatity(opt_types, strike, spot, 0.03, 0.01, time_to_maturity, settle)      # 计算隐含波动率
    # 返回的是dataframe格式，要转化为数值型
    calculated_vol = calculated_vol.iloc[0,0]
    # 期权的时间价值为负时(此种情况在50ETF期权里时有发生)，没法通过BSM模型计算隐含波动率，故此时将期权隐含波动率设为0.0，实际上，此时的隐含波动率和各风险指标并无实际参考价值
     # 但是为了防止分母为0，故设置为0.01%
    if calculated_vol is None:
        calculated_vol = 0.0001
    return  calculated_vol   
```

```python
import pandas as pd

TradingDaysSeries=data["tradeDate"]
TradingDaysList=[]
for i in TradingDaysSeries:
    TradingDaysList.append(i)
TradingDaysList=list(set(TradingDaysList))
TradingDaysList=sorted(TradingDaysList)
del TradingDaysSeries

Holding=pd.DataFrame();


for i in TradingDaysList:
    Trimming=data[data["tradeDate"]==i]
    
    ExpDateSeries=Trimming["expDate"]
    ExpDateList = []
    for j in ExpDateSeries:
        ExpDateList.append(j)    
    ExpDateList=list(set(ExpDateList))
    ExpDateList=sorted(ExpDateList)    
    CurrentMonth=ExpDateList[1]
    #选出当日的近月平值看涨期权，并计算出隐含波动率、delta值
    
    TrimmingTemp1=Trimming[Trimming["expDate"]==CurrentMonth][Trimming["contractType"]=="PO"]
    TrimmingTemp1Sort=TrimmingTemp1.sort(columns=["diff"])
    TrimmingTemp1Sort=TrimmingTemp1Sort[2:3]
    
    print TrimmingTemp1Sort
    
    #print TrimmingTemp1Sort[0:1]
    TrimmingTemp1Sort.loc[:,'delta']=option_delta(TrimmingTemp1Sort)
    TrimmingTemp1Sort.loc[:,'option_iv']=option_iv(TrimmingTemp1Sort)
    #print TrimmingTemp1Sort[0:1]
    print TrimmingTemp1Sort
    
    Holding=Holding.append(TrimmingTemp1Sort[0:1])

Holding.head(20)    
    
```
