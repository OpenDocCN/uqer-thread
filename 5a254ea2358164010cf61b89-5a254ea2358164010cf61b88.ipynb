# “交割宝”策略

参考方正证券的一篇研报，策略思路如下：
	只在交割周进行交易，也就是说一个月就搞一次。
    每次选在交割周的星期一。
    根据历史50ETF数据算出85%置信区间的3日最大波动。
    筛选出基本不可能交割的深度虚值期权卖出，一直持有到到期。

策略逻辑比较简单，但是唯一的缺点就是遇到黑天鹅事件就比较残，**谨慎谨慎**。
后续可以改进的地方，就是根据历史数据计算的3日最大波幅，如果最近一段时间波动比较剧烈，说明市场不稳定，此时卖出期权并不明智。

```python
import pandas as pd
import numpy as np
import time
import matplotlib.pyplot as plt
```

```python
tradeDay = [20150223, 20150323, 20150427, 20150525, 20150622, 20150727, 20150824, 20150928, 20151026, 20151123, 20151228, 20160125, 20160222, 20160328, 20160425, 20160523, 20160627, 20160725, 20160822, 20160926, 20161024, 20161128, 20161226, 20170123, 20170227, 20170327, 20170424, 20170522, 20170626, 20170724, 20170828, 20170925, 20171023, 20171127, 20171225]
tradeDay = map(lambda x:str(x)[:4]+'-'+str(x)[4:6]+'-'+str(x)[-2:],tradeDay)
```

```python
# 手动导入的ETF数据
ETF_data = pd.read_csv('50ETF.csv',index_col=0)
ETF_data.columns = ['close','high','low','pre_close','open']
#ETF_data.index = map(lambda x: time.strftime("%Y-%m-%d",time.strptime(x,"%Y/%m/%d")),ETF_data.index)
```

```python
ETF_data['change'] = ETF_data['close']-ETF_data['pre_close']
ETF_data['3dayupdown'] = ETF_data['change'].rolling(3,min_periods=3).sum()

ETF_data['85percen'] = pd.rolling_apply(ETF_data['3dayupdown'], 200,lambda x: np.percentile(x,85))
ETF_data['15percen'] = pd.rolling_apply(ETF_data['3dayupdown'], 200,lambda x: np.percentile(x,15))
```

```python
netvalue = pd.Series(index = ETF_data.index, data = np.zeros_like(ETF_data.index))
```

```python
def get_option_prices(day,secID,field):
    df = DataAPI.MktOptdGet(tradeDate=day,
                            secID=secID,optID=u"",ticker=u"",
                            beginDate=u"",endDate=u"",field=field,pandas="1")
    return df[field][0]
```

```python
for i in range(len(ETF_data.index)):
    day = ETF_data.index[i]
    if day in tradeDay:
        openprice = ETF_data.ix[day, 'open']
        upband = openprice + ETF_data.ix[day, '85percen']
        downband = openprice + ETF_data.ix[day, '15percen']
        optionData = DataAPI.OptDpoGet(secID=u"",optID=u"",ticker=u"",varSecID=u"",
                             varTicker=u"510050",varType=u"",beginDate=day,endDate=day,exerType=u"",
                             contractType=u"",field=u"",pandas="1")
        optionData['lastTradeDa'] = map(lambda x : x[:-3],optionData['lastTradeDate'])
        optionData = optionData[optionData['lastTradeDa']==day[:-3]]
        Calloption = optionData[optionData['contractType']=='CO']
        Putoption = optionData[optionData['contractType']=='PO']

        Calloption = Calloption[Calloption['strikePrice']>=upband]
        Putoption = Putoption[Putoption['strikePrice']<=downband]

        optionData = pd.concat([Calloption,Putoption],axis=0)
        if len(optionData)>0:            
            Select_secID = optionData['secID'].tolist()
            ETF_price = ETF_data.ix[ETF_data.index[i+2],'close']
            df = pd.DataFrame(index = Select_secID,columns=['firstOpen','fclose','sclose','tclose','isstrike'],
                              data=np.zeros((len(Select_secID),5)))
            df['firstOpen'] = map(lambda x:get_option_prices(day,x,'openPrice'),Select_secID)
            df['fclose'] = map(lambda x:get_option_prices(day,x,'closePrice'),Select_secID)
            df['sclose'] = map(lambda x:get_option_prices(ETF_data.index[i+1],x,'closePrice'),Select_secID)
            df['tclose'] = map(lambda x:get_option_prices(ETF_data.index[i+2],x,'closePrice'),Select_secID)
            for sec in df.index:
                if sec[6]=='P':
                    df.ix[sec,'isstrike'] = -max(int(sec.split('.')[0][-4:])/1000.0-ETF_price,0)
                elif sec[6]=='C':
                    df.ix[sec,'isstrike'] = -max(ETF_price-int(sec.split('.')[0][-4:])/1000.0,0)
            
            netvalue[day] = np.sum(-(df['fclose']-df['firstOpen']))*10000
            netvalue[ETF_data.index[i+1]] = np.sum(-(df['sclose']-df['fclose']))*10000
            netvalue[ETF_data.index[i+2]] = np.sum(-(2*df['tclose']-df['sclose']))*10000+np.sum(df['isstrike'])*10000
        #print day
```

```python
netvalue.cumsum().plot()
plt.grid()
```
