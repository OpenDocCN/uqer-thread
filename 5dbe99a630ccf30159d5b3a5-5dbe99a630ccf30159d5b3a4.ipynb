# 概率角度的投资-条件概率

## 条件概率

```python
import numpy as np
import pandas as pd
import datetime
import time
import matplotlib.pyplot as plt
```

```python
##股票选取
def chs_factor(ticker = '000010',begin = '20010101',end = '20170201', field = ['tradeDate' ,'closePrice', 'preClosePrice']):
    dataday = DataAPI.MktEqudAdjGet(tradeDate=u"",secID=u"",ticker=ticker,beginDate= begin,endDate= end,isOpen="1",field= field,pandas="1")
    dataday['chgPct'] = dataday['closePrice'] / dataday['preClosePrice'] - 1
    return dataday
```

```python
df_ori = chs_factor()
## 作为后面的基准概率
base_rate = df_ori['chgPct'].apply(lambda x: 0 if x < 0 else 1).mean()
print(base_rate)
```

以一只股票为例，我们从概率的角度来看今天股票会不会涨，用概率来看就可以描述成$p(\text{up})$。
在没有信息的条件下，可以通过统计来计算该概率。如下：
就是说，通过历史信息的统计，我们可以知道这只股票的上涨概率是0.539。
但是，这种方法太过于粗糙，我们是可以得到其他信息的，但这种统计并没有计算其他信息。

```python
stat = df_ori['chgPct'].apply(lambda x: 0 if x < 0 else 1)
stat_dict = {}
for i in range(stat.shape[0] - 1):
    st = (stat.values[i], stat.values[i + 1])
    if st not in stat_dict.keys():
        stat_dict[st] = 1
    stat_dict[st] += 1

stat_1 = float(stat_dict[(0,1)] + stat_dict[(1, 1)])
print(stat_dict[(0,1)] / stat_1, stat_dict[(1,1)] / stat_1)
```

那我们对上涨的描述应该是$p(\text{up} | info)$，即在我们已知一些信息的条件下，我们描述上涨的概率应该是一个条件概率。
一个简单的例子就是，在我已知前一天涨的情况下，今天上涨的概率是多少。在我已知前一天跌的情况下，上涨的概率又是多少。
再往后推导，我们知道的信息并不是只有一天，而是很多天呢。这个概率就可以往后拓展。

```python
stat = df_ori['chgPct'].apply(lambda x: 0 if x < 0 else 1)
delay_day = 4
stat_dict = {}
for i in range(stat.shape[0] - delay_day):
    st = tuple(stat.values[i + j] for j in range(delay_day))
    if st not in stat_dict.keys():
        stat_dict[st] = 1
    stat_dict[st] += 1
up_index = [key for key in stat_dict.keys() if key[-1] == 1]
stat_sum = np.sum([stat_dict[key] for key in up_index]).astype(np.float)
for key in up_index:
    print(key, ':',stat_dict[key] / stat_sum)
```

```python
sigals = []
for i in range(stat.shape[0] - (delay_day - 1)):
    st = tuple(stat.values[i + j] for j in range(delay_day - 1))
    if st == (1,1,1):
        sigals.append(1)
    else:
        sigals.append(0)
```

```python
df_ori['signal'] = [0]*(delay_day - 1) +  sigals
```

如果我们只做一种投资方式，那就是买最可能涨的模式，从上面的推论可以看出来来，（1，1，1，1）是最可能涨的模式，因为其条件概率最高。如何买到（1，1，1，1）模式呢，我们能做的就是当出现（1，1，1）时，进行投资就可能拿到（1，1，1，1）这种频繁模式，投资到（1，1，1，1）这种模式。

```python
plt.figure(figsize= (10, 5))
(df_ori['chgPct']+ 1).cumprod().plot(label = 'ori_ser')
(df_ori['chgPct'] * df_ori['signal'] + 1).cumprod().plot(label = 'stratage')
plt.legend()
```

更进一步，我们重新思考。我们的投资决策应该是，$p(\text{up} | info) > p( \text{down} | info)$

```python
stat = df_ori['chgPct'].apply(lambda x: 0 if x < 0 else 1)
delay_day = 5

# 统计模式
stat_dict = {}
for i in range(stat.shape[0] - delay_day):
    st = tuple(stat.values[i + j] for j in range(delay_day))
    if st not in stat_dict.keys():
        stat_dict[st] = 1
    stat_dict[st] += 1
    
# 统计前模式
stat_pre_dict = {}
for i in range(stat.shape[0] - delay_day + 1):
    st = tuple(stat.values[i + j] for j in range(delay_day - 1))
    if st not in stat_pre_dict.keys():
        stat_pre_dict[st] = 1
    stat_pre_dict[st] += 1

# 统计概率
result_dict = {}
for pre_key in stat_pre_dict.keys():
    stat_sum = np.float(stat_pre_dict[pre_key])
    result_dict[pre_key] = {}
    for key in stat_dict.keys():
        if pre_key == key[:-1]:
            result_dict[pre_key][key[-1]] = stat_dict[key] / stat_sum
```

```python
up_index = [key for key in result_dict.keys() if result_dict[key][1] > base_rate]
sigals = []
for i in range(stat.shape[0] - (delay_day - 1)):
    st = tuple(stat.values[i + j] for j in range(delay_day - 1))
    if st in up_index:
        sigals.append(1)
    else:
        sigals.append(0)
df_ori['signal'] = [0]*(delay_day - 1) +  sigals
```

```python
## 训练集上的表现
plt.figure(figsize= (10, 5))
(df_ori['chgPct']+ 1).cumprod().plot(label = 'ori_ser')
(df_ori['chgPct'] * df_ori['signal'] + 1).cumprod().plot(label = 'stratage')
plt.legend()
```
