下面是一些简单的Hurst指数和R/S方法的介绍。
基于重标极差(R/S)分析方法基础上的赫斯特指数(H)的研究是由英国水文专家H．E．Hurst(1900—1978)在研究尼罗河水库水流量和贮存能力的关系时，发现用有偏的随机游走(分形布朗运动)能够更好地描述水库的长期存贮能力，并在此基础上提出了用重标极差(R/S)分析方法来建立赫斯特指数(H)。作为判断时间序列数据遵从随机游走还是有偏的随机游走过程的指标。
    洪水过程是时间系列曲线，具有正的长时间相关效应。即干旱愈久，就可能出现持续的干旱；大洪水年过后仍然会有较大洪水。这种特性可以用赫斯特指数来表示。
    赫斯特指数有三种形式：
    1．如果H=0.5，表明时间序列不存在长期记忆性；
    2．如果0.5<H<1，表明时间序列存在长期记忆性；
    3．如果0≤H<0.5，表明时间序列有反持续性，为均值回复过程。
    其实Hurst指数分析属于分型市场分析，可以归集于非线性经济学，是对有效市场理论得一种补充
    将Hurst指数运用到中国A股市场的研究有很多，大致有一些共识：一个就是A股市场是存在长期记忆性的，即Hurst指数绝大部分时候是超过0.5的，第二个是目前A股上证综指的平均循环周期长度是三百多个交易日。
    本次研究的主要目的有两个：
    一是确定上证综指的周期；
    二是以此周期为时间窗口，计算滚动Hurst指数，并和上证指数进行一定的对比分析

    计算的数据对象为上证综指（选取的时间段是重1997年1月之后的，因为在1996年12月，我国开始实行涨跌停板制度），先是计算出上证综指的对数收益率，然后再通过AR(1)对收益率进行处理，得出残差序列，最终对残差序列进行处理，这是为了消除短期的依赖。

```python
    R／S分析法是十分有效的工具，不必假定潜在的分布是高斯分布，仅仅独立就可以。这样R／S分析法的研究对象不仅包括正台分布，而且也包括非高斯独立过程，如t、F及其他分布。R／S分析法是非参数分析方法，这样对于所讨论的分布的形状就没有什么特别要求。对于一个系统来说，H=1／2意味着一个独立过程。但当H=l／2，却并不能说明时问序列是一个高斯随机游走，仅表明不存在长期记忆。
    R／S分析的基本思想是：
    1、对一个长度为T的时间序列{X_t}，把它分为A个长度为n的等长子区间，因而T=A*n。标记每个子区问为I_a，a=1,2，3，…，A。在子区间I_a中，每一元素标记为N(k,a)，k=1,2，3，，n。长度为n的子区间平均值被定义为：
    u_a=mean(I_a)
    2、作为每一个子区间I_a，对于均值的累积离差定义如下（以向量形式表示）：
    第一步，先求离差：y = X_t - p*u_a                  #n行，a列，对每一列求离差
    第二步，求累计离差：sum = np.cumsum(y,axis=0)          #n行，a列，求累计离差    
    3、极差被定义为在每一个子区间I_t，累计离差的最大值减去累计离差的最小值
    r = np.max(sum,axis=0) - np.min(sum,axis=0)      #%1行，a列
    4、对于每一个子区间I_a的样本标准差：
    sig = np.std(X_t,axis=0)             #1行，a列，对每一列求标准差
    5、每一个极差r，都对应于它的标准差s。这样一来，对A个长度为n的相邻子区问。我们定义如下：
    RS[n] = np.mean(r / sig,axis=0)        #1行，1列
    这是每个子序列长度为n时候的RS值
    6、对于不同的划分方法，重复步骤(I)一(5)我们都可以得到不同的RS_n。Hurst通过长时问的实践总结．建立了如下关系
    R／S=c*pow(n,H)
    对此式两边取对数
    log(R／s) = H*log(n)+log(c) 其中C为常数
    因此，对log(n)和log(R/S)进行最小二乘法回归就可以估计出H的值
    下面的统计量V是由Hurst(1951)用于检验R／S分析的稳定性的，但也可以很好地用来估计周期长度。
    V[n] = RS[n] * (pow(n,-0.5))
    如果R／S以时间的平方根标度，那么这个比例V_n是一条水平线。换句话说，对于独立的随机过程来说，统计量V_n关于Iog(n)是平坦的。另一方面，对于具有状态持续性的过程来说，R／S以大于时间的平方根的标度(H>0.5)，V_n关丁log(n)向上倾斜；反之，对于具有反持续(H<0.5)的过程来说，V_n关于log(n)向下倾斜。当V_n图形状态发生改变时，就产生突变，长期记忆过程消失。
    下面首先进行V统计量的计算，通过寻找V统计量的极大值来判断出目前的平均循环周期   
    因为选取的是实行涨停板制度之后数据，共计有4千8百多条，所以子序列长度取在10到480的范围内，因为根据历史经验和大量的论文，子序列长度小于10，估计值会不稳定，而且V统计量出现极值点的时候子序列长度基本在三百多，所以取10到480的范围能能够看到V统计量的极值点，这是从实用的角度出发的
```

```python
#X    #需要处理的时间序列
#T    #时间序列的总长度（4千8百多）
#A    #子序列个数，
#n    #子序列长度（10-480）
#a    #子序列个数
#RS   R/S指数
#V    指数

RS = np.zeros(480)      #array
RS1 = np.zeros(480)      #array
RS2 = np.zeros(480)      #array
V = np.zeros(480)       #array
logRS = np.zeros(480)   #array
logn = np.zeros(480)   #array
X = np.array(X)    #读入时间序列,一维矩阵
T = X.shape[0]     #读取时间序列的总长度，
for n in xrange(10,480):  #每个长度计算一次
    a = int(T/n) 
    x1 = X[0:n*a].reshape(n,a)   #正向分段,n行a列，每一列为一个子序列
    x2 = X[(T - a * n): T].reshape(n,a)  #反向分段,n行a列，每一列为一个子序列
    m1 = np.mean(x1,axis=0)     #按列取均值，1行a列
    m2 = np.mean(x2,axis=0)     #按列取均值，1行a列
    p = np.ones((n,1))              #n行，1列
    y1 = x1 - p * m1                  #n行，a列，对每一列求离差
    y2 = x2 - p * m2                  #n行，a列，对每一列求离差       
    sig1 = np.std(x1,axis=0)             #1行，a列，对每一列求标准差
    sig2 = np.std(x2,axis=0)             #1行，a列，对每一列求标准差
    sum1 = np.cumsum(y1,axis=0)          #n行，a列，求累计离差
    sum2 = np.cumsum(y2,axis=0)          #n行，a列，求累计离差
    r1 = np.max(sum1,axis=0) - np.min(sum1,axis=0)      #%1行，a列
    r2 = np.max(sum2,axis=0) - np.min(sum2,axis=0)      #%1行，a列        
    RS1[n] = np.mean(r1 / sig1,axis=0)        #1行，1列
    RS2[n] = np.mean(r2 / sig2,axis=0)        #1行，1列    
    RS[n] = 0.5*RS1[n] + 0.5*RS2[n]
    logRS[n] = np.log(RS[n])
    V[n] = RS[n] * (pow(n,-0.5))
    logn[n] = np.log(n)
R = regression.linear_model.OLS(logRS[10:480],logn[10:480]).fit()
R.params[0]
```

    V统计量在318的时候出现最大值，姑且可以认为上证综指有一个平均循环周期，长度为318个交易日，所以我们接下来进行滚动Hurst指数计算的时候，滚动窗口定为38日

```python
def Hurst(XX,T):  
    XX = np.array(XX)    #读入时间序列,一维矩阵
    Lenth = XX.shape[0]     #读取时间序列的总长度，
    hurst = np.zeros(Lenth)
    for i in xrange(T,Lenth):
        X=XX[i-T:i+1]
        RS = np.zeros(T)      #array
        logRS = np.zeros(T)   #array
        logn = np.zeros(T)   #array        
        for n in xrange(10,T):  #每个长度计算一次
            a = int(T/n) 
            x1 = X[0:n*a].reshape(n,a)   #正向分段,n行a列，每一列为一个子序列
            x2 = X[(T - a * n): T].reshape(n,a)  #反向分段,n行a列，每一列为一个子序列
            m1 = np.mean(x1,axis=0)     #按列取均值，1行a列
            m2 = np.mean(x2,axis=0)     #按列取均值，1行a列
            p = np.ones((n,1))              #n行，1列
            y1 = x1 - p * m1                  #n行，a列，对每一列求离差
            y2 = x2 - p * m2                  #n行，a列，对每一列求离差       
            sig1 = np.std(x1,axis=0)             #1行，a列，对每一列求标准差
            sig2 = np.std(x2,axis=0)             #1行，a列，对每一列求标准差
            sum1 = np.cumsum(y1,axis=0)          #n行，a列，求累计离差
            sum2 = np.cumsum(y2,axis=0)          #n行，a列，求累计离差
            r1 = np.max(sum1,axis=0) - np.min(sum1,axis=0)      #%1行，a列
            r2 = np.max(sum2,axis=0) - np.min(sum2,axis=0)      #%1行，a列        
            RS1[n] = np.mean(r1 / sig1,axis=0)        #1行，1列
            RS2[n] = np.mean(r2 / sig2,axis=0)        #1行，1列    
            RS[n] = 0.5*RS1[n] + 0.5*RS2[n]
            logRS[n] = np.log(RS[n])
            logn[n] = np.log(n)
        R = regression.linear_model.OLS(logRS[10:T],logn[10:T]).fit()
        hurst[i] = R.params[0]
    return hurst
```

时间窗口取为318，从第318个交易日开始，每个交易日都计算出一个Hurst指数的值，这个值是以该交易日和之前的318个交易日为基础算出来的。

下面是一个简单的策略，该策略调仓的依据是Hurst指数下穿0.5374线，该线是Hurst指数的期望平均线，Hurst指数下穿平均线，可以认为一段趋势即将结束，进入反转趋势。在此条件下进行再细分，加仓的依据是大盘指数在60日均线下方，可以认为大盘处于低位，要进入反转上升，所以加仓；减仓依据依据是大盘指数在60日均线上方，可以认为大盘处于高位，如果反转将进入熊市，所以减仓，甚至空仓。

一、回测详情
1、通过回测，可以发现该策略在2007年牛市的时候能够有效跟随，但是反应慢于牛市启动时间，所以牛市时候收益率跑输给基准，但是在2008年熊市来临的时候能够及时卖出，保住一定的利润，当然卖出时间也是晚于熊市来临。此处有一个地方比较遗憾就是2015年的牛市，该策略未能抓住牛市，而是出于空仓。
2、该策略空仓时间比较长，该段时间可以投资流动性较高收益稳定的其他金融资产。
3、该策略调仓次数较少，十年调仓10次左右，所以回测时候没有考虑手续费模块。
二、下一步改进的地方
1、继续完善细节，抓住牛市启动，不能错过大的牛市，这样至少能作为一个仓位控制的参考。
