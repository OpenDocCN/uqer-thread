4.3.3数据探索与清洗
通常我们会认为ROE的值约为10%-30%，但是实际上在调用所有股票的ROE因子进行分析时，我们发现并不是这么一回事。以ROE为例，介绍常用的数据探索方法及数据清洗的方式。
我们通常在进行探索数据时，通常会想通过做一个简单的直方图来看数据的分布。正式的ROE分布可能与我们想象的并不一样，存在许多负值。并且存在-3——3的极端值。


```python
df=DataAPI.MktStockFactorsOneDayGet(tradeDate=u"20180119",secID=set_universe("A"),ticker=u"",
                                    field=u"secID,tradeDate,ROE",pandas="1")
df.plot.hist(bins=100)
```

可以再做一下boxplot图来看看结果。

```python
_=df.boxplot(sym='rs')
```

如上图所示，在数据中存在很多的异常值。当然这些异常值按照传统算法，可能并不算是错误值，但是如果要把ROE当做一个指标，进一步分析其对股票未来收益或其他方面的影响，则在建立回归或者是其他模型时就必须要考虑到对异常值的处理，因为他们对模型的影响可能很大。
这里我们以最常用的3倍标准差法为例，将超过3倍标准差的数据调整为3倍标准差

```python
df_=df.copy()
```

```python
df_.loc[df.ROE-df.ROE.mean()<-3*df.ROE.std(),'ROE']=df.ROE.mean()-3*df.ROE.std()
df_.loc[df.ROE-df.ROE.mean()>-3*df.ROE.std(),'ROE']=df.ROE.mean()+3*df.ROE.std()
```

```python
df_.plot.hist(bins=100)
```

4.3.4数据转换
1.数据标准化。
数据标准化可以标准化可以消除量纲及变量自身变异程度的影响。常用的标准化方法有Min-max标准化和z-score标准化
Min-max标准化计算的公式为：v'=(v-min)/(max-min)    标准化后的值统一到0—1
z-score标准化的计算公式为：z'=(v-u)/sigma          标准化后的数据分布为均值为0，方差为1

```python
((df_['ROE']-df_['ROE'].min())/(df_['ROE'].max()-df_['ROE'].min())).plot.hist(bins=100)
```

```python
((df_['ROE']-df_['ROE'].mean())/df_['ROE'].std()).plot.hist(bins=100)
#标准化不会改变数据原始序数，他做的知识对数值大小的调整。若不看值的话，则数据的分布并没有变化
```
