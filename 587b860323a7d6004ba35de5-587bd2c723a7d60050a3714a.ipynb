### 1. Dual Thrust策略
Dual Thrust策略是一种趋势跟踪系统，属于日内交易策略。该策略由Michael Chalek 在20世纪80年代开发，曾被Future Thruth杂志评为最赚钱的策略之一。Dual Thrust系统具有简单易用和适用度广的特点，其思路简单且参数少，配合不同的参数、止盈止损和仓位管理可以为投资者带来长期稳定的收益。而且该策略适用品种较多，被投资者广泛应用于股票、货币、贵金属、债券、能源及股指期货市场等。在Dual Thrust交易系统中，对于震荡区间的定义非常关键，这也是该交易系统的核心。Dual Thrust在Range的设置上，引入前N日的四个价位，Range = Max(HH-LC,HC-LL)来描述震荡区间的大小。其中HH 是N日High的最高价，LC是N日Close的最低价，HC是N日Close的最高价，LL是N日Low的最低价。这种方法使得一定时期内的Range相对稳定，可以适用于日间的趋势跟踪。Dual Thrust对于多头和空头的触发条件，考虑了非对称的幅度，做多和做空参考的Range可以选择不同的周期数，也可以通过参数K1和K2来确定。具体分为两步来实现：

第一步：计算相关参数，得到上轨Buyline 和下轨Sellline：
* N日High的最高价HH, N日Close的最低价LC
* N日Close的最高价HC，N日Low的最低价LL
* Range = Max(HH-LC,HC-LL)
* BuyLine = Open + K1\*Range
* SellLine = Open - K2\*Range

第二步：交易逻辑：
* 当价格向上突破上轨时，如果当时持有空仓，则先平仓，再开多仓；如果没有仓位，则直接开多仓；
* 当价格向下突破下轨时，如果当时持有多仓，则先平仓，再开空仓；如果没有仓位，则直接开空仓；

当K1时，多头相对容易被触发,当K1>K2时，空头相对容易被触发。因此，投资者在使用该策略时，一方面可以参考历史数据测试的最优参数，另一方面，则可以根据自己对后势的判断，或从其他大周期的技术指标入手，阶段性地动态调整K1和K2的值。

### 2. 函数预备
datacenter ： 存储日内数据
getDayRange: 计算Range
report: 计算平均盈利/平均亏损等

```python
class DataCenter(object):
    """
    self.refresh_rate ： 获取数据的周期
    self.pipe_length : 存储数据的长度
    self.data: 存储需要间隔的数据
    self.symbol： 数据所属合约
    self.coef ： 合约每手的单位
    """
    keys = [u'openPrice', u'highPrice', u'lowPrice', u'closePrice', u'tradeDate', u'volume']
    def __init__(self,account, refresh_rate, pipe_length):
        self.data = {key:deque([], maxlen=pipe_length) for key in self.keys}
        self.refresh_rate = refresh_rate
        self.pipe_length = pipe_length
        self.symbol=None
        self.account = account
        self.date = None
        self.coef=None
    def update_data(self, symbol):
        hist = get_symbol_history(symbol=symbol, field=self.keys,time_range=self.refresh_rate)[symbol]
        if self.symbol != symbol:
            self.clear()
            self.symbol = symbol
            self.coef = np.array(DataAPI.FutuGet(ticker=symbol,field=u"contMultNum",pandas="1").contMultNum)[-1]
        if self.date != self.account.current_date:
            self.clear()
            self.date = self.account.current_date
        self.data['closePrice'].append(hist.ix[-1, 'closePrice'])
        self.data['openPrice'].append(hist.ix[0, 'openPrice'])
        self.data['highPrice'].append(hist['highPrice'].max())
        self.data['lowPrice'].append(hist['lowPrice'].min())
        self.data['volume'].append(hist['volume'].sum())
        self.data['tradeDate'].append(hist.ix[len(hist)-1, 'tradeDate'])
        return self.data
    def clear(self):
        for i in self.keys:
            self.data[i].clear()
```

```python
def getDayRange(data,n): 
    """
    data: 日间数据
    n: 计算range的周期
    """
    hh = np.array(data['highestPrice'])[-n:].max()
    hc = np.array(data['closePrice'])[-n:].max()
    lc = np.array(data['closePrice'])[-n:].min()
    ll = np.array(data['lowestPrice'])[-n:].min()
    return max(hh-lc,hc-ll)
```

```python
def report(pref):
    returns = np.array(list(pref['returns']))
    win_streak =''
    for i in range(len(returns)):
        if returns[i] > 0:
            win_streak = win_streak + '1'
        if returns[i] < 0:
            win_streak = win_streak + '0'
    max_win_streak = 0
    max_lose_streak = 0
    for i in win_streak.split('0'):
        if len(i)>max_win_streak:
            max_win_streak = len(i)
    for i in win_streak.split('1'):
        if len(i)>max_lose_streak:
            max_lose_streak = len(i)
    return {'平均盈利/平均亏损': np.mean(returns[returns>0])/float(abs(np.mean(returns[returns<0]))),
            '总盈利/总亏损':sum(returns[returns>0])/float(abs(sum(returns[returns<0]))),
            '胜率':len(returns[returns>0])/float(len(returns[returns!=0])),
            '交易次数': len(returns[returns!=0]),
            '最大连胜次数': max_win_streak,
            '最大连败次数': max_lose_streak 
           }
```

### 3. 回测主体

止损方法是买入价cost亏损比例stop_rate或盈利一定比例win_rate,则平仓。
不采用我在[《整理形态之上升三角形》](https://uqer.datayes.com/community/share/58723de023a7d6004ba35bfd)中提出的动态调整的方法，因为日内数据变化较小。

```python
def backtest(universe,start,end,k1,k2,stop_rate, win_rate,n=4,capital_base = 1e6,refresh_rate = 5,freq = 'm',RK=None ):
    """
    universe :策略期货合约
    start: 回测开始时间
    end  : 回测结束时间
    capital_base : 初试可用资金
    refresh_rate: 调仓周期
    freq : 调仓频率,m-> 分钟；d-> 日
    k1 : 上轨参数
    k2 : 下轨参数
    """
    commission = {'IF':(0.0025/100.0, 'perValue'), 
                  'IH':(0.0025/100.0, 'perValue'), 
                  'RB': (0.0045/100.0, 'perValue'),
                  'CU': (0.005/100.0, 'perValue'),
                  'AU': (10, 'perShare'),
                  'J': (0.008/100.0, 'perValue'),
                  'JM': (0.01/100.0, 'perValue')}
    def initialize(futures_account):           # 初始化虚拟期货账户，一般用于设置计数器，回测辅助变量等。
        global data_pool
        data_pool = DataCenter(futures_account,refresh_rate, 8)
        futures_account.range= np.NAN
        futures_account.buyline= np.NAN
        futures_account.sellline= np.NAN
        futures_account.barsinceentry = 0
        futures_account.entryprice = np.NAN

    def handle_data(futures_account):          # 回测调仓逻辑，每个调仓周期运行一次，可在此函数内实现信号生产，生成调仓指令。
        symbol = get_symbol(universe)
        if futures_account.current_time == '09:15:00':
            end = futures_account.previous_date
            calendar = Calendar('China.SSE')
            begin = calendar.advanceDate(end,'-'+str(n+3)+'B').toDateTime().date()
            prevData = DataAPI.MktFutdGet(ticker=symbol, beginDate=begin,endDate=end,
                                           field=[u"closePrice", u"lowestPrice",u"openPrice",u"highestPrice"], pandas="1")
            prevData = prevData.dropna(axis=0)
            futures_account.range = getDayRange(prevData,n)
            today_open = DataAPI.MktFutdGet(tradeDate=futures_account.current_date, ticker=symbol, field=[u"openPrice"], pandas="1")
            open_price = today_open.ix[0, 'openPrice']
            futures_account.buyline = open_price + k1*futures_account.range
            futures_account.sellline = open_price - k2*futures_account.range
        elif futures_account.current_time >= '09:15:00' and futures_account.current_time <= '14:30:00':
            long_position = futures_account.position.get(symbol, dict()).get('long_position', 0)
            short_position = futures_account.position.get(symbol, dict()).get('short_position', 0)
            data = data_pool.update_data(symbol)
            coef = data_pool.coef
            close_price = np.array(data['closePrice'])
            if len(close_price)<2: return
            open_price = np.array(data['openPrice'])
            low_price = np.array(data['lowPrice'])
            high_price = np.array(data['highPrice'])
            
            if close_price[-2] < futures_account.buyline and close_price[-1] > futures_account.buyline:
                amount = futures_account.portfolio_value/(coef*close_price[-1])
                if short_position != 0:
                    order(symbol, short_position, 'close')
                    futures_account.barsinceentry = 0
                if long_position == 0:
                    order(symbol, amount, 'open')
                    futures_account.barsinceentry = 1
            if close_price[-2] > futures_account.sellline and close_price[-1] < futures_account.sellline:
                amount = futures_account.portfolio_value/(coef*close_price[-1])
                if long_position != 0:
                    order(symbol, -long_position, 'close')
                    futures_account.barsinceentry = 0
                if short_position == 0:
                    order(symbol, amount, 'open')
                    futures_account.barsinceentry = 1

            if futures_account.barsinceentry == 2:
                futures_account.entryprice = open_price[-1]
            if futures_account.barsinceentry != 0:
                futures_account.barsinceentry += 1
            if long_position != 0:
                if (close_price[-1]/futures_account.entryprice < 1-stop_rate 
                    or close_price[-1]/futures_account.entryprice > 1+win_rate):
                    order(symbol, -long_position, 'close')
                    futures_account.barsinceentry = 0
                    
            if short_position != 0:
                if (close_price[-1]/futures_account.entryprice > 1+stop_rate 
                    or close_price[-1]/futures_account.entryprice < 1-win_rate):
                    order(symbol, short_position, 'close')
                    futures_account.barsinceentry = 0
                    
        elif futures_account.current_time >= '14:50:00':
            long_position = futures_account.position.get(symbol, dict()).get('long_position', 0)
            short_position = futures_account.position.get(symbol, dict()).get('short_position', 0)
            futures_account.barsinceentry = 0
            futures_account.entryprice = np.NAN
            if long_position != 0:
                order(symbol, -long_position, 'close')
            if short_position != 0:
                order(symbol, short_position, 'close')
                
        else:
            return
    bt, perf = qf.backtest.backtest(universe=universe, start=start, end=end,
                                initialize=initialize, handle_data=handle_data,
                                capital_base=capital_base, refresh_rate=refresh_rate,
                                freq=freq,commission = commission )
    return perf
```

### 4. K值的选取
K值太大，则很难突破，即便突破，可能也接近闭市，这时，盈利的可能性极大减少。
当K值越小，信号虽然可能产生较早，但是可能有很多虚假信号。
在[ 《国内商品期货常用日内CTA策略测试》](https://uqer.datayes.com/community/share/586f319989e3ba0047efdc09)一文中，k值选取值大概在0.5附近，该文中利用的是一日的Range，可以理解为，大概半日内便可确定是否交易。

以上仅限于讨论，无实际指导指标
下面先从一日的Range确定最佳k系数。

```python
start = '20160203'
end   = '20160630'
universe = 'RBM0'

k = np.round(np.linspace(0.3, 1.2, 10),2).tolist()

annualized_return = pd.DataFrame(index=k,columns=['k1','k2'])
max_drawdown = pd.DataFrame(index=k,columns=['k1','k2'])
sharpe = pd.DataFrame(index=k,columns=['k1','k2'])
average_PL_ratio = pd.DataFrame(index=k,columns=['k1','k2'])
total_PL_ratio = pd.DataFrame(index=k,columns=['k1','k2'])
Ptimes_ratio = pd.DataFrame(index=k,columns=['k1','k2'])
traded_times = pd.DataFrame(index=k,columns=['k1','k2'])
max_win_streak  = pd.DataFrame(index=k,columns=['k1','k2'])
max_lose_streak = pd.DataFrame(index=k,columns=['k1','k2'])

for para in k:

    re1=backtest(universe,start,end,para, 10,.05, .12,n=1)#k2很大为10，则对k1测试无影响。
    re2=backtest(universe,start,end,10, para,.05, .12,n=1)
    
    report1= report(re1)
    report2= report(re2)
    
    annualized_return.ix[para,'k1'] = np.float64(re1['annualized_return'])
    annualized_return.ix[para,'k2'] = np.float64(re2['annualized_return'])
    max_drawdown.ix[para,'k1'] = np.float64(re1['max_drawdown'])
    max_drawdown.ix[para,'k2'] = np.float64(re2['max_drawdown'])
    sharpe.ix[para,'k1'] = np.float64(re1['sharpe'])
    sharpe.ix[para,'k2'] = np.float64(re2['sharpe'])
    average_PL_ratio.ix[para,'k1'] = np.float64(report1['平均盈利/平均亏损'])
    average_PL_ratio.ix[para,'k2'] = np.float64(report2['平均盈利/平均亏损'])
    total_PL_ratio.ix[para,'k1'] = np.float64(report1['总盈利/总亏损'])
    total_PL_ratio.ix[para,'k2'] = np.float64(report2['总盈利/总亏损'])
    Ptimes_ratio.ix[para,'k1'] = np.float64(report1['胜率'])
    Ptimes_ratio.ix[para,'k2'] = np.float64(report2['胜率'])
    traded_times.ix[para,'k1'] = np.float64(report1['交易次数'])
    traded_times.ix[para,'k2'] = np.float64(report2['交易次数'])
    max_win_streak.ix[para,'k1'] = np.float64(report1['最大连胜次数'])
    max_win_streak.ix[para,'k2'] = np.float64(report2['最大连胜次数'])
    max_lose_streak.ix[para,'k1'] = np.float64(report1['最大连败次数'])
    max_lose_streak.ix[para,'k2'] = np.float64(report2['最大连败次数'])
    print 'para = %.2f：回测结束'%para
    print '----------------------------------------------------------------'
```

### 5.不同的N值

选15日、4日、1日有什么区别？ 这个很重要。
** 首先假想一下区别 **
1. N值太小，可以尽快反应波动，但Range相比相邻时间Range变化可能较快
2. N值太大，Range较为稳定，但很难反应波动
3. N值适中，Range较为稳定，避免快速波动；而且有'大趋势大突破，小趋势小突破'
4. 在[ 《国内商品期货常用日内CTA策略测试》](https://uqer.datayes.com/community/share/586f319989e3ba0047efdc09)一文中提出的ATR过滤对于一日的Range有些用处。如果N大于1，在一定程度上起着相同的作用。
