# 配对交易作业

作业要求：

原理：一种基于模型的投资过程，在不依赖于经济含义的情况下，运用数量手段构建资产组合，根据证券价格与数量模型所预测的理论价值进行对比，构建证券投资组合的多头和空头，从而对市场风险进行规避，获取一个稳定的超额收益率。

难点：选择配对的股票。统计套利是一种风险套利。

作业目标： 在A股市场上，选择相关性较强的配对股票，在买入一支的同时卖空另外一支，构造初始时刻投入资金为0的股票交易策略。并根据设定的阈值进行投资或者平仓。在一定时期内，统计该策略的收益及风险。


作业注意事项：
    1. 考虑实际股票市场佣金、印花税、过户费三项费用，买入卖出的交易费税取为千分之一。 
    2. 此作业为个人作业，需要上交作业报告。

可进一步改进和思考的地方：
    1.	该策略模型有哪些改进方法，预期改进结果如何？

实现思路：

1. 策略前的研究。
	a. 获取申万行业分类，手动寻找同质性强的行业，并记录
    b. 获取记录的行业中的股票，获取这些股票过去一年的收盘价
    c. 对这些股票收盘价进行协整检验，记录通过协整检验的股票对。
    d. 对通过协整检验的股票对历史收盘价进行线性回归，对回归结果进行筛选，记录通过回归检验的股票对

3. 策略
	a. 根据之前研究的方法，确定股票对获取函数
	b. 在较长的一段时间内内对筛选出的股票对进行投资，初始资金100W。
	c. 投资方法为：（开仓信号阈值，提前平仓）参数进行投资。
	d. 投资m个月后，回到a重新选取股票对。

```python
import pandas as pd
import numpy as np
from CAL.PyCAL import *
import datetime
pd.set_option('display.width', 200)
import statsmodels.api as sm
import seaborn as sns
import matplotlib.pyplot as plt

ids = {'industry_name' : ['bank' , 'insurance' , 'highway' , 'hydropower' , 'flavoring' , 'Cement' , 'aluminum' , 'tungsten' , 'General_steel' , 'Coal_mining' , 'Coke_processing' , 'nitrogenous_fertilizer' , 'polyester_fiber'] , 'industryID' : ['010303210101' , '010303220201' , '010303190201' , '010303180102' , '010303140102' , '010303060101' , '010303050101' , '010303050302' , '010303040101' , '010303020201' , '010303020202' , '010303030301' , '010303030401']}
industry = pd.DataFrame(ids)

def find_cointegrated_pairs(dataframe):
    # 得到DataFrame长度
    n = dataframe.shape[1]
    # 初始化p值矩阵
    pvalue_matrix = np.zeros((n, n))
    # 抽取列的名称
    keys = dataframe.keys()
    # 初始化强协整组
    pairs = pd.DataFrame(columns=['stock1' , 'stock2' , 'coint_pvalue'])
    # 对于每一个i
    for i in range(n):
        # 对于大于i的j
        for j in range(i+1, n):
            # 获取相应的两只股票的价格Series
            stock1 = dataframe[keys[i]]
            stock2 = dataframe[keys[j]]
            # 分析它们的协整关系
            result = sm.tsa.stattools.coint(stock1, stock2)
            # 取出并记录p值
            pvalue = result[1]
            pvalue_matrix[i, j] = pvalue
            if pvalue < 0.01:
                # 记录股票对和相应的p值
                p = pd.DataFrame([[keys[i], keys[j], pvalue]] , columns=['stock1' , 'stock2' , 'coint_pvalue'])
                pairs = pairs.append(p)
    # 返回结果
    return pvalue_matrix , pairs

#获取这些行业中的所有股票
std_in_industry = DataAPI.EquIndustryGet(industry=u"",secID=u"",ticker=u"",industryVersionCD=u"010303",intoDate=u"",industryID1=u"",industryID2=u"",industryID3 = industry['industryID'].tolist(),field= ['secID' , 'secShortName', 'industryID3' , 'isNew'] , pandas="1")
std_in_industry = std_in_industry[std_in_industry['isNew']==1].reset_index(drop=True)
#print std_in_industry

#获取每个行业中所有股票半年内的收盘价，并进行协整检验 ， 记录所有通过协整检验的股票对
pairs_all = pd.DataFrame(columns=['stock1' , 'stock2' , 'coint_pvalue'])
for i in range(industry.shape[0]):
    std = std_in_industry[std_in_industry['industryID3'] == industry['industryID'][i]]
    
    closeprice = pd.DataFrame()
    stock_list = std['secID'].tolist()
    for j in stock_list:
        cp = DataAPI.MktEqudGet(secID= j ,ticker=u"",tradeDate=u"",beginDate=u"20160101",endDate=u"20170101",isOpen="1",field=['closePrice'],pandas="1")
        if cp.empty:
            continue
        else:
            cp.dropna(axis=1, how='any', thresh= 2*len(cp)/3, subset=None, inplace=False)
            cp.rename(columns={'closePrice' : j}, inplace = True)
            closeprice = pd.concat([closeprice, cp], axis=1, ignore_index=False)
    #print closeprice
    pvalues, pairs= find_cointegrated_pairs(closeprice)
    #sns.set()
    #print pvalues
    print industry['industryID'][i] , industry['industry_name'][i] , len(closeprice.keys()),pairs.shape[0]
    plt.figure()
    ax = sns.heatmap(pvalues,xticklabels= closeprice.keys() , yticklabels= closeprice.keys() , cmap='RdYlGn_r', mask = (pvalues == 0))
    plt.title(industry['industry_name'][i])
    plt.show()
    pairs = pairs.reset_index(drop=True)
    print pairs
    if pairs.shape[0] >=1:
        pairs_all = pairs_all.append(pairs)
pairs_all = pairs_all.reset_index(drop=True)
print pairs_all
pairs_all.to_csv('协整股票对/coint_pairs.csv')
```

对协整股票对进行线性回归，根据线性回归的结果筛选合适的股票对

```python
import pandas as pd
import numpy as np
from CAL.PyCAL import *
import datetime
pd.set_option('display.width', 200)
import statsmodels.api as sm
import seaborn as sns
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import adfuller
import scipy.stats as scs

pairs_all = pd.read_csv('协整股票对/coint_pairs.csv')
def zscore(series):
    return (series - series.mean()) / np.std(series)

def show_pairs(pairs):
    for i in range(pairs.shape[0]):
        plt.figure(figsize=(25,5))
        x = DataAPI.MktEqudGet(secID = pairs['stock1'][i] ,ticker=u"",tradeDate=u"",beginDate=u"20160101",endDate=u"20170101",isOpen="1",field=['closePrice'],pandas="1")
        y = DataAPI.MktEqudGet(secID = pairs['stock2'][i] ,ticker=u"",tradeDate=u"",beginDate=u"20160101",endDate=u"20170101",isOpen="1",field=['closePrice'],pandas="1")
        x.rename(columns={'closePrice' : pairs['stock1'][i]}, inplace = True)
        y.rename(columns={'closePrice' : pairs['stock2'][i]}, inplace = True)
        X = sm.add_constant(x)
        result = (sm.OLS(y,X)).fit()
        print result.summary()
        plt.subplot(1 , 4 , 1)
        plt.plot(x); plt.plot(y)
        plt.xlabel("Time"); plt.ylabel("Price")
        plt.legend([pairs['stock1'][i], pairs['stock2'][i]],loc='best')

        plt.subplot(1 , 4 , 2)
        #fig, ax = plt.subplots(figsize=(8,6))
        plt.plot(x, y, 'o', label="data")
        plt.plot(x, result.fittedvalues, 'r', label="OLS")
        plt.legend(loc='best')

        plt.subplot(1 , 4 , 3)
        plt.plot(zscore(result.params[pairs['stock1'][i]]*x.sub(y[pairs['stock2'][i]],axis=0)))
        plt.axhline(zscore(result.params[pairs['stock1'][i]]*x.sub(y[pairs['stock2'][i]],axis=0)).mean()[pairs['stock1'][i]] , color="black")
        plt.axhline(1.0, color="red", linestyle="--")
        plt.axhline(-1.0, color="green", linestyle="--")
        plt.legend(["z-score", "mean", "+1", "-1"])
        plt.xlabel("Time"); plt.ylabel("Stationary Series")
        plt.legend(["z-score", "mean", "+1", "-1"])
        
        plt.subplot(1 , 4 , 4)
        plt.hist(result.resid, bins=5)

pairs = pd.DataFrame(columns=['stock1' , 'stock2' ])
 


#获取筛选出来的股票对最近半年的收盘价，并进行线性回归
for i in range(pairs_all.shape[0]):
    x = DataAPI.MktEqudGet(secID = pairs_all['stock1'][i] ,ticker=u"",tradeDate=u"",beginDate=u"20160101",endDate=u"20170101",isOpen="1",field=['closePrice'],pandas="1")
    y = DataAPI.MktEqudGet(secID = pairs_all['stock2'][i] ,ticker=u"",tradeDate=u"",beginDate=u"20160101",endDate=u"20170101",isOpen="1",field=['closePrice'],pandas="1")
    x.rename(columns={'closePrice' : pairs_all['stock1'][i]}, inplace = True)
    y.rename(columns={'closePrice' : pairs_all['stock2'][i]}, inplace = True)
    

    #线性回归
    X = sm.add_constant(x)
    result = (sm.OLS(y,X)).fit()
    #print(result.summary())
    #print result.params
    #print result.pvalues
    
    t_pairs = pd.DataFrame(columns=['stock1' , 'stock2'])
    t_pairs = pd.DataFrame([[pairs_all['stock1'][i] , pairs_all['stock2'][i]]] , columns = ['stock1' , 'stock2'])
    
    #plt.figure(figsize=(25,10))
    if result.pvalues['const'] < 0.1 and result.pvalues[pairs_all['stock1'][i]]:
        if result.rsquared_adj > 0.8:
            if adfuller(result.resid)[1] < 0.01:
                if abs(scs.skew(result.resid)) < 0.4:
                    pairs = pairs.append(t_pairs)

                
pairs = pairs.drop_duplicates(subset = ['stock1'] , keep = 'first')
pairs = pairs.drop_duplicates(subset = ['stock2'] , keep = 'first')
pairs.reset_index(drop = True , inplace = True)

show_pairs(pairs)

print pairs.values.flatten()
```
