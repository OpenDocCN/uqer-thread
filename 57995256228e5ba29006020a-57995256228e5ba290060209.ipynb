script.dir <- dirname(sys.frame(1)$ofile)
setwd(script.dir)
source("MCOptionTools.R")
TypeFlag <- "p"; S = 100; X = 90; Time = 1/12*90/30; sigma = 0.4; b = 0.03; r = 0.03; delta.t = 1/245

ret <- MCOption(TypeFlag = TypeFlag, S = S, X= X,Time = Time, sigma = sigma, b = b, r = r, delta.t = delta.t,
                mcLoops = 100, pathLength = 245*Time, mcSteps = 10000, init = TRUE,
                innovations.gen = sobolInnovations, path.gen = wienerPath, 
                payoff.calc = plainVanillaPayoff, scrambling = 2, seed=4711)

ret2 <- MCOption(TypeFlag = TypeFlag, S = S, X= X,Time = Time, sigma = sigma, b = b, r = r, delta.t = delta.t,
                 mcLoops = 100, pathLength = 245*Time, mcSteps = 10000, init = TRUE,
                 innovations.gen = sobolInnovations, path.gen = wienerPath, 
                 payoff.calc = plainVanillaPayoff.delta, scrambling = 2, seed=4711)

ret3 <- GBSOption(TypeFlag, S, X, Time, r, b, sigma)@price
cat("\n")
print(paste("BS价", ret3))
print(paste("MC价", mean(ret)))
print(paste("Delta价", mean(ret2)))
