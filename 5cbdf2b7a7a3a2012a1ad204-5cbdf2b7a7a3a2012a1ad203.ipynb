```python
# 常见指数
generalindex = pd.read_csv('generalindex.csv',encoding='utf-8',header=None)

generalindex.columns = ['SecID','指数简称','发布日期','指数类别']
generalindex.iloc[0,0] = '000002.ZICN'
generalindex['ticker'] = generalindex.SecID.str.slice(0,6)

generalindex = generalindex.set_index('SecID')
generalindex = generalindex.sort_index(ascending=True)
print (generalindex.shape)
print (generalindex.loc['000001.ZICN']['指数简称'])
# generalindex = generalindex.reset_index()
print (generalindex.to_html())
```

```python
# 获取指数的成分股
index_detail = DataAPI.IdxConsGet(secID='000002.ZICN',ticker=u"",isNew=u"",pandas="1")
print (index_detail.head().to_html())
```

```python
# 获取各指数对应的成分股
uqer_generalindex = []
index_consid_list = []

for secid in generalindex.index:
    
    try:  
        
        index_detail = DataAPI.IdxConsGet(secID=secid,ticker=u"",isNew=u"",pandas="1")
        print index_detail['secShortName'][0],secid,len(index_detail)
        uqer_generalindex.append(secid)
        index_detail = index_detail.rename(columns={'consID':secid})
        index_consid_list.append(index_detail[secid])
        
        index_info = DataAPI.IdxGet(secID=secid,ticker=u"",field=u'',pandas="1")
        index_info['tickersNO'] = len(index_detail)
        columns = ['secShortName','secID','tickersNO','publishDate','ticker','indexType','baseDate','porgFullName']
        index_info = index_info[columns]
        # print (index_info.to_html()) 
    except: 
        continue

print (len(uqer_generalindex))

index_consid = pd.concat(index_consid_list,axis=1,join='outer')
print (index_consid.head().to_html())
```

```python
# 查找某个公司包含在那些指数里面
secid = '600012.XSHG'
for c in index_consid.columns:
    if secid in index_consid.loc[:,c].tolist():
        print (generalindex.loc[c]['指数简称'])
```

```python
# 各指数成分股交集情况
index_relation = pd.DataFrame(index = index_consid.columns,columns = index_consid.columns)

for c in index_relation.columns:
    for r in index_relation.index:
        cr_set = set(index_consid[c]) & set(index_consid[r])
        if c == '000001.ZICN' or r == '000001.ZICN':
            index_relation.loc[c,r] = len(cr_set)
        else:
            index_relation.loc[c,r] = len(cr_set) - 1

dict1 = dict(zip(generalindex.index,generalindex['指数简称']))
print (len(dict1))
# import json
# print json.dumps(dict1, encoding="UTF-8", ensure_ascii=False)
index_relation = index_relation.rename(columns=dict1) 
index_relation = index_relation.rename(dict1) 
print (index_relation.to_html())
```
