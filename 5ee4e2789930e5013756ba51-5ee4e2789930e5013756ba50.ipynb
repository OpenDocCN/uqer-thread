```python
'''
研究目的
1.寻找一段时间内，连续3天上涨，均线发散（ma5>ma10>ma20)的股票
2.成交量逐渐放大
3.观察一段时间后的走势
'''
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime
ss='20200428'
ee='20200612'

from CAL.PyCAL import *
cal = Calendar('CHINA.SSE')
a=cal.bizDatesNumber(ss, ee, includeFirst=True, includeLast=True)
# print(a)

f=['secID','secShortName','listDate','totalShares']
pool=DataAPI.EquGet(secID=u"",ticker=u"",equTypeCD=u"A",listStatusCD=u"L",exchangeCD="",ListSectorCD=u"",field=f,\
                    pandas="1")
print(pool.tail())
#剔除新股，ss-listdate>100，至少上市超过100个交易日
#要把ss和listDate全部转为datetime类型

ss1=datetime.datetime.strptime(ss,'%Y%m%d')
print(ss1)
pool['listDate']=pd.to_datetime(pool['listDate'])
# print(pool.dtypes)
# print(len(pool))
test=pool['listDate'][0]
tmp=(ss1-test)



for i in range(len(pool)):
    tmp=pool['listDate'][i]
    shares=pool['totalShares'][i]
    if (ss1-tmp).days<=120 or shares>=1000000000:#总股本小于50亿股，还能炒炒吧
        pool=pool.drop([i])

pool=pool.reset_index(drop=True)

print(len(pool))

result={'id':[],'date':[],'secName':[]}
ff=['secID','tradeDate','secShortName','closePrice','turnoverVol']
# testpool=['002197.XSHE','300413.XSHE','002163.XSHE']

for i in pool['secID']:
    rawdata=DataAPI.MktEqudGet(secID=i,ticker=u"",tradeDate=u"",beginDate=ss,endDate=ee,isOpen="1",\
                           field=ff,pandas="1")
    close=rawdata['closePrice']
    # print(close)
    
    # print(name)
    t=rawdata['turnoverVol']
    iid=rawdata['secID']
    for j in range(len(close)-2):
            if close[j+2]>close[j+1]>close[j]:
                if t[j+2]>t[j+1]>t[j]:
                    result.get('id').append(i)
                    result.get('date').append(rawdata['tradeDate'][j])
                    result.get('secName').append(rawdata['secShortName'][j])
                    break

result=pd.DataFrame(result)
print(result)
```

```python

ff=['secID','tradeDate','EMA5','EMA10','EMA20']
# result=result[-10:]
# result=result.reset_index(drop=True)
# print(result)
finalout={'id':[],'date':[]}
for i in range (len(result)):
    tmp=DataAPI.MktStockFactorsOneDayGet(tradeDate=result['date'][i],secID=result['id'][i],\
                                         ticker=u"",field=ff,pandas="1")
    t5=tmp['EMA5']
    t10=tmp['EMA10']
    t20=tmp['EMA20']
    # print(tmp)
    if t5[0]>t10[0]>t20[0]:
        finalout.get('id').append(result['id'][i])
        finalout.get('date').append(result['date'][i])
finalout=pd.DataFrame(finalout)
print(finalout)

ft=['secID','tradeDate','closePrice','turnoverRate']

out={'date':[],'id':[],'pct':[],'turnoverRate':[]}

for i in range (len(finalout)):
    
    ti=DataAPI.MktEqudGet(secID=finalout['id'][i],ticker=u"",tradeDate=u"",beginDate=finalout['date'][i],\
                      endDate=ee,isOpen="",field=ft,pandas="1")
    # print(ti)
    tmpclose1=ti['closePrice'][0]
    tmpclose2=ti['closePrice'][len(ti)-1]
    p=(tmpclose2-tmpclose1)/tmpclose1
    out.get('date').append(finalout['date'][i])
    out.get('id').append(finalout['id'][i])
    out.get('pct').append(p)
    out.get('turnoverRate').append(ti['turnoverRate'][2])
out=pd.DataFrame(out)
print(out)
```
