```python
def test_stationarity(timeseries):
    
    #滑动均值和方差
    rolmean = timeseries.rolling(12).mean()
    rolstd = timeseries.rolling(12).std()

    #绘制滑动统计量
    plt.figure(figsize=(24, 8))   
    orig = plt.plot(timeseries, color='blue',label='Original')
    mean = plt.plot(rolmean, color='red', label='Rolling Mean')
    std = plt.plot(rolstd, color='black', label = 'Rolling Std')
    
    plt.legend(loc='best')
    plt.title('Rolling Mean & Standard Deviation')
    plt.show()
    
    #adf检验
    print('Results of Dickey-Fuller Test:')
    dftest = adfuller(timeseries, maxlag =12,regression = 'c',autolag='AIC')
    print (dftest)
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used',
                                             'Number of Observations Used'])
    for key,value in dftest[4].items():
        dfoutput['Critical Value (%s)'%key] = value
    print(dfoutput)
test_stationarity(SZMDATA['closePrice'])
```

```python
# car_output_data['output'] = SZMDATA['closePrice']
# car_output_data['output'] = car_output_data['output'].astype('float64')
car_diff = SZMDATA['closePrice'].diff(1)
car_diff =  car_diff.dropna()
print car_diff.head(10)
test_stationarity(car_diff)
car_diff = car_diff.astype('float64')
```

```python
car_diff.values
```

```python
plot_acf(car_diff,lags=36)
plot_pacf(car_diff,lags=36)
plt.show()
```

```python
r,rac,Q = sm.tsa.acf(car_diff, qstat=True)
prac = pacf(car_diff,method='ywmle')
table_data = np.c_[range(1,len(r)), r[1:],rac,prac[1:len(rac)+1],Q]
table = pd.DataFrame(table_data, columns=['lag', "AC","Q", "PAC", "Prob(>Q)"])
print table
```

```python
#(p, q) =(sm.tsa.arma_order_select_ic(car_diff,max_ar=3,max_ma=3,ic='aic')['aic_min_order'])
order = sm.tsa.arma_order_select_ic(car_diff,max_ar=3,max_ma=3,ic=['aic', 'bic', 'hqic'])
order.bic_min_order
```

```python
#ARMA  模型预测
# model = ARIMA(car_diff, order=(3,1,2))

# model_fit = model.fit(disp=0)

# print(model_fit.summary())

mod = sm.tsa.statespace.SARIMAX(car_diff, trend='n', order=(3,2,2), seasonal_order=(1,1,1,12))
results = mod.fit()
print(results.summary())
results.predict(start = 350, end= 354, dynamic= True)  
```

```python
#ARMA  模型预测
MODEL20=ARMA(car_diff,(3,2)).fit()
resid = MODEL20.resid
fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot(111)
fig = qqplot(resid, line='q', ax=ax, fit=True)
plt.show()
summary = (MODEL20.summary2(alpha=.05, float_format="%.8f"))
print(summary)
```

```python
t=sm.tsa.stattools.adfuller(resid)
output=pd.DataFrame(index=['Test Statistic Value', "p-value", "Lags Used", "Number of Observations Used","Critical Value(1%)","Critical Value(5%)","Critical Value(10%)"],columns=['value'])
output['value']['Test Statistic Value'] = t[0]
output['value']['p-value'] = t[1]
output['value']['Lags Used'] = t[2]
output['value']['Number of Observations Used'] = t[3]
output['value']['Critical Value(1%)'] = t[4]['1%']
output['value']['Critical Value(5%)'] = t[4]['5%']
output['value']['Critical Value(10%)'] = t[4]['10%']
print(output)
```
