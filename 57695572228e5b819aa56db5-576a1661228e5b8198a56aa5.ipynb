```python
import datetime
import re
import pandas as pd
import numpy as np
from lib import dectofrac

date_end = datetime.datetime.now().strftime("%Y%m%d")
date_start = datetime.datetime.now() - datetime.timedelta(days=30)
date_start = date_start.strftime("%Y%m%d")

Futu = DataAPI.FutuGet(exchangeCD=u"",secID=u"",ticker=u"",contractStatus="L",contractObject=u"",field=u"",pandas="1")
tikers = Futu['ticker']
p=re.compile('[a-zA-Z]+')
cons = []
tickers=tikers.tolist()
for ticker in tickers:
    a=p.match(ticker)
    if len(cons)==0 or cons[-1]!=a.group():
        cons.append(a.group())
        
mainCons = []
for con in cons:
    sids = DataAPI.FutuGet(exchangeCD=u"",secID=u"",ticker=u"",contractStatus="L",contractObject=con,field=u"",pandas="1")['secID']
    data = {}
    for sid in sids :
        one = DataAPI.MktFutdGet(secID=sid,beginDate=date_start,endDate=date_end,field= 'mainCon')
        if not one['mainCon'].empty and one['mainCon'][len(one['mainCon'])-1]:
            mainCons.append(sid)
            
for sid in mainCons :
    one = DataAPI.MktFutdGet(secID=sid,beginDate=date_start,endDate=date_end,field= 'closePrice')
    data.update({sid : one['closePrice']})
        
p = pd.DataFrame(data) 
corrs = p.corr()

n=0
sid_list = []
for sid0 in mainCons:
    sid_list.append(sid0)
    if sid0 in corrs:
        corrs = corr[sid0]
        for (sid1,cor) in zip(corr.columns.tolist(),corrs.tolist()):
            if 1 > cor > 0.9 and sid1 not in sid_list:
                p1 = np.polyfit(data[sid1],data[sid0] , 1)
                n += 1
                factor1 = np.poly1d(p1)
                print 'Index : '+str(n)
                print ' : '.join([sid1,sid0])               
                print ' : '.join(['cor',str(cor),str(factor1)])
                vol0 = DataAPI.FutuGet(exchangeCD=u"",secID=sid0,ticker=u"",contractStatus="L",contractObject=u"",field=u"",pandas="1")['contMultNum']
                vol1 = DataAPI.FutuGet(exchangeCD=u"",secID=sid1,ticker=u"",contractStatus="L",contractObject=u"",field=u"",pandas="1")['contMultNum']
                print ' : '.join(['vol',str(vol0[0]),str(vol1[0])])
                frac = p1[0]*vol0[0]/vol1[0]
                frac = dectofrac.decimal_to_fraction(frac,20)
                print ' : '.join(['frac',str(frac)])
                print ' '

```
