```python
def wrapper_func_cache(func):
    @wraps(func)
    def wrapped(*args, **kwargs):
        if not hasattr(func, '__cache_result'):
            func.__cache_result = {}
        
        hash_key = [str(i) for i in args] + [str(j) for j in kwargs.iteritems()]
        cache_key = hash(tuple(hash_key))
        
        if cache_key not in func.__cache_result:
            func.__cache_result[cache_key] = func(*args, **kwargs)
            
        return func.__cache_result[cache_key]
    return wrapped

def trade_range(tstart, tend):
    data = DataAPI.TradeCalGet(exchangeCD=u"XSHG", beginDate=tstart, endDate=tend, field=['calendarDate', 'isOpen'])
    return list(data[data.isOpen == 1]['calendarDate'])
```

```python
### 最大回撤归因
def drawdown_details(bt, perf):
    """根据累计价值计算最大回撤，返回最大回撤日期序列
    start, end, drawdown, cumulative_values
    """
    res = pd.DataFrame()
    start, end, drawdown = [], [], []
    cumulative_values = list(perf['cumulative_values'])
    for index in range(len(bt)):
        today_value = cumulative_values[index]
        today = bt.tradeDate.iloc[index]
        max_pre_value = max(cumulative_values[: index + 1])   # 之前的最高价值，包括当天
        max_pre_value_day = bt.tradeDate.iloc[cumulative_values.index(max_pre_value)]
        _drawdown = 1 - today_value / max(1, max_pre_value)
        start.append(max_pre_value_day)
        end.append(today)
        drawdown.append(_drawdown)
    res['start'], res['end'], res['drawdown'], res['cumulative_values'] = start, end, drawdown, cumulative_values
    res['drawdown_period'] = ['{} -> {}'.format(str(i)[:10], str(j)[:10]) for i, j in zip(list(start), list(end))]
    
    return res

def attribution_drawdown(bt, perf):
    """回撤分布
    """
    drawdown = drawdown_details(bt, perf)
    describe = drawdown.describe()
    describe = describe[describe.index.isin([u'mean', u'min', u'25%', u'50%', u'75%', u'max'])]
    term, value = [], []
    
    ### 回撤
    term.extend([i+j for i in ['最大回撤 | '] for j in ['平均值', '最小', '25分位', '50分位', '75分位', '最大']])
    value.extend(list(describe['drawdown']))
    
    ### 最大回撤区间
    max_drawdown = value[-1]
    index = list(drawdown['drawdown']).index(max_drawdown)
    max_drawdown_period = list(drawdown.drawdown_period)[index]
    term.append('最大回撤发生区间')
    value.append(max_drawdown_period)

    return term, value, drawdown
```

```python
def attribution_returns(perf):
    """日收益详情归因
    """
    returns = perf['returns']
    daily_returns = pd.DataFrame()
    daily_returns['tradeDate'], daily_returns['returns'] = list(returns.index), list(returns.values)
    daily_returns['benchmark'] = list(perf['benchmark_returns'])
    describe = daily_returns.describe()
    describe = describe[describe.index.isin([u'mean', u'min', u'25%', u'50%', u'75%', u'max'])]
    term, value = [], []
    ### 日收益四分位值
    term.extend([i+j for i in ['每日收益率 | '] for j in ['平均值', '最小', '25分位', '50分位', '75分位', '最大']])
    value.extend(list(describe['returns']))
    
    return daily_returns, term, value
```

```python
def attribution_frequency(blotter, bt, perf):
    """交易频率分析
    """
    trade_days = sorted(set(blotter.order_time.str.slice(0, 11, 1)))
    backtest_range = list(perf['benchmark_cumulative_returns'].index)
    tstart, tend = str(backtest_range[0])[:10], str(backtest_range[-1])[:10]
    full_trade_days = trade_range(tstart, tend)
    term = ['回测时间', '回测期间总交易日', '策略交易日', '交易频率']
    value = ['{} -> {}'.format(tstart, tend), len(full_trade_days), len(trade_days), round(len(trade_days)*1./len(full_trade_days), 2)]
    
    return term, value
    
def attribution_order_status(blotter):
    """订单状态
    """
    state_counter = collections.Counter(list(blotter.state))
    term, value = [], []
    state_dict = {'UpLimit': '涨停', 'DownLimit': '跌停', 'Suspended': '停牌', 'Normal': '[全部/部分/未]成交'}
    for i in state_dict:
        state_counter[i] += 0
        
    total = sum(state_counter.values())
    for i in state_counter:
        term.append('订单状态: ' + state_dict[i])
        value.append('{:<4} | {}'.format(state_counter[i], round(state_counter[i]*1./total, 3)))

    return term, value

def attribution_ticker_distribution(blotter):
    """证券分布
    """
    tickers = collections.Counter(list(blotter.symbol))
    total = sum(tickers.values())
    term, value = [], []
    term.extend(['交易证券支数', '交易次数'])
    value.extend([len(tickers), total])
    for _index, (_sec, _number) in enumerate(tickers.most_common(10)):
        term.append('交易次数排名：{}'.format(_index))
        value.append('{} | {:3} | {:.3f}'.format(_sec, _number, round(_number*1./total, 4)))
    
    return term, value
```

```python
def extract_order(record):
    """解析一条买/卖订单
    原始：Order(order_time: 2016-01-04 09:30, symbol: 600099.XSHG, direction: 1, order_amount: 2200, state: Normal, 
               filled_time: 2016-01-04 09:30, filled_amount: 2200, transact_price: 13.6700, slippage: 0.0000, commission: 0.0137)
    返回：[   ['order_time', ' 2016-01-04 09:30'],
             ['symbol', '600099.XSHG'],
             ['direction', '1'],
             ['order_amount', '2200'],
             ['state', 'Normal'],
             ['filled_time', ' 2016-01-04 09:30'],
             ['filled_amount', '2200'],
             ['transact_price', '13.6700'],
             ['slippage', '0.0000'],
             ['commission', '0.0137']
        ]
    """
    str_order = str(record)
    list_order = str_order[str_order.index('(') + 1 : str_order.index(')')].split(",")
    list_order = [i.strip() for i in list_order]
    
    def str2record(_str):
        """'order_time: 2013-01-08 09:30'  -> ['order_time', '2013-01-08 09:30']
        """
        if _str.startswith('order_time') or _str.startswith('filled_time'):
            return [_str[0 : _str.index(":")], _str[_str.index(":") + 1 : ]]
        else:
            return [i.strip() for i in _str.split(':')]
    
    list_order = [str2record(i) for i in list_order]
    
    return list_order

def extract_blotter(blotter):
    """解析下单记录，从 order 类解析成一条完整的买卖记录
    """
    all_records = {'order_time': [], 'symbol': [], 'direction': [], 'order_amount': [], 'state': [],
                   'filled_time': [], 'filled_amount': [], 'transact_price': []}
    for basket in blotter:
        for i in basket:
            tmp = extract_order(i)
            for term, value in tmp:
                if term in all_records:
                    all_records[term].append(value)
    
    blotter = pd.DataFrame.from_records(all_records)
    blotter.direction = blotter.direction.astype(float)
    blotter.filled_amount = blotter.filled_amount.astype(float)
    blotter.transact_price = blotter.transact_price.astype(float)
    blotter.order_amount = blotter.order_amount.astype(float)
    
    return blotter

def attribution(bt, perf):
    """回测归因报告
    
    return: 订单，简要报告，调仓行情[ticker, buy, sell, hold_days, earning, eraning_rate, market_value, pe, pb]，回撤统计，每日收益
    """
    report = pd.DataFrame(columns=['项目', '数据'])
    detail_report = pd.DataFrame(columns=['项目', '数据'])
    term, value = [], []
    
    # 自带回测指标
    keys = ['annualized_return', 'benchmark_annualized_return', 'alpha', 'beta', 'sharpe', 'volatility', 
            'information_ratio', 'max_drawdown', 'turnover_rate']
    performance = {'annualized_return': "年化收益率", 'benchmark_annualized_return': "基准年化收益率", 
                   'alpha': "阿尔法", 'beta': '贝塔', 'sharpe': '夏普比率', 'volatility': '收益波动率', 
                   'information_ratio': '信息比率', 'max_drawdown': '最大回撤', 'turnover_rate': '换手率'}
    for k in keys:
        term.append(performance[k])
        value.append(perf[k])
    # print u'解析回测变量 bt 和 perf'
    # 解析回测变量 bt 和 perf
    blotter = extract_blotter(bt.blotter)
    
    # 交易频率
    # print u'交易频率'
    _term, _value = attribution_frequency(blotter, bt, perf)
    term.extend(_term)
    value.extend(_value)
    
    # 回撤分布
    # print u'回撤分布'
    _term, _value, drawdown = attribution_drawdown(bt, perf)
    term.extend(_term)
    value.extend(_value)

    # 日收益分布
    # print u'日收益分布'
    daily_returns, _term, _value = attribution_returns(perf)
    term.extend(_term)
    value.extend(_value)
    
    # 订单状态
    # print u'订单状态'
    _term, _value = attribution_order_status(blotter)
    term.extend(_term)
    value.extend(_value)
   
    # 证券分布
    # print u'证券分布'
    _term, _value = attribution_ticker_distribution(blotter)
    term.extend(_term)
    value.extend(_value)
    
    
    # 回测报表
    for i in range(len(value)):
        if isinstance(value[i], (int, float, np.float)):
            value[i] = '{:.4f}'.format(value[i])
    report['项目'], report['数据'] = term, value    
    
    return blotter, report, drawdown, daily_returns
```

```python
############################################
### graph
############################################

def fig_style(ax, legend, legend_loc='upper right', xlabel=''):
    legend = [i.decode('utf-8') if type(i) == str else i for i in legend]
    ax.legend(legend, prop=small_font, loc=legend_loc)
    ax.title.set_font_properties(font)
    if xlabel:
        ax.set_xlabel(xlabel, fontproperties=small_font)
    ax.set_axis_bgcolor('white')
    ax.grid(color='gray', alpha=0.2, axis='y')
    return ax

def plot_hist(data_dict):
    """画直方分布图
    """
    import numpy as np
    import matplotlib.pyplot as plt 
    
    heights, bins = [], []
    fig, ax = plt.subplots(figsize=(20,6))
    color_set = ['cornflowerblue', 'seagreen', 'yellow', 'blue', 'gray']
    
    for index, column in enumerate(data_dict.keys()):
        if len(heights) and len(bins):
            heights, bins = np.histogram(data_dict[column], bins=bins)
            width = (bins[1] - bins[0]) / (len(data_dict) + 1)
            ax.bar(bins[ : -1] + width * index, heights, width=width, facecolor=color_set[index])
        else:
            heights, bins = np.histogram(data_dict[column], bins=20)
            width = (bins[1] - bins[0]) / (len(data_dict) + 1)
            ax.bar(bins[ : -1], heights, width=width, facecolor=color_set[index])
        
    ax = fig_style(ax, legend=data_dict.keys(), legend_loc='best')
    return ax
```

```python
class StrategyPool(object):
    """策略池，多策略分析
    """
    def __init__(self):
        self.strategy_pool = {}        # key: strategy_name; value: (bt, perf)
        self.report = pd.DataFrame()
        self.returns = pd.DataFrame()
        self.cache = {}
    
    def register(self, name, bt, perf):
        """加入一个策略到策略池
        """
        self.strategy_pool[name] = (bt, perf)
    
    def items(self):
        """打印当前策略名字
        """
        for i in self.strategy_pool:
            print i
    
    def clear(self):
        """清空策略池
        """
        self.strategy_pool = {}
    
    def backtest_graph(self):
        """绘制策略池回测曲线图；注意对齐回测始末时间。
        """
        for strategy in self.strategy_pool:
            _bt, _perf = self.strategy_pool[strategy]
            if 'tradeDate' not in self.returns:
                self.returns['tradeDate'] = list(_bt.tradeDate)
                self.returns['benchmark'] = list(_perf['benchmark_cumulative_returns'])
            self.returns[strategy] = list(_perf['cumulative_returns'])
        
        ### plot
        x = u'tradeDate'
        y = [i for i in self.returns.columns if i != u'tradeDate']
        ax = self.returns.plot(x=x, y=y, figsize=(22, 10), title=u'多策略回测对比', linewidth=3)
        fig_style(ax, legend=[i.decode('utf-8') for i in y], legend_loc='best', xlabel=u'交易日')
    
    def attribution_table(self):
        """策略池各策略归因表
        """
        for strategy in self.strategy_pool:
            self.cache[strategy] = {} 
            _bt, _perf = self.strategy_pool[strategy]
            blotter, report, drawdown, daily_returns = attribution(_bt, _perf)
            self.cache[strategy]['blotter'] = blotter
            self.cache[strategy]['report'] = report
            self.cache[strategy]['drawdown'] = drawdown
            self.cache[strategy]['daily_returns'] = daily_returns
            
            if '项目' not in self.report:
                self.report['项目'] = report['项目']
            self.report[strategy] = report['数据']
        
        print self.report.to_html()
```
