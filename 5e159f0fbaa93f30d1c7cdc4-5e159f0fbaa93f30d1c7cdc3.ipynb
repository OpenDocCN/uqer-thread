```python
import pandas as pd
import numpy as np

def clc_nav(nav):
    '''
    功能：给定净值序列，计算各组合的评价指标，包括：年化收益率、年化标准差、夏普值、最大回撤
    输入：
        nav  series，index为升序排列的日期，columns为各标的代码，value为每日收盘价
    '''
    nav_pct = nav.pct_change().dropna()
    res = pd.Series()
    res['annRet'] = round((nav.values[-1]/nav.values[0])**(250.0/len(nav.values))-1,3)
    res['sharp'] = res['annRet'] / (np.std(nav.pct_change().dropna()) * np.sqrt(250))
    res['maxDrawdown'] = abs(min([nav.values[i:].min() / nav.values[i] - 1 for i in range(len(nav.values))]))
    res['calmarRatio'] = res['sharp']/res['maxDrawdown']
    res['winRate'] = (nav_pct>0).sum()/len(nav_pct)
    res['maxWin'] = nav_pct.max()*100
    res['maxLoss'] = nav_pct.min()*100
    res['rangeDay'] = len(nav)
    res['meanRate'] = nav_pct[nav_pct>0].mean()*100/abs(nav_pct[nav_pct<0].mean()*100)
    res['weekRet'] = round(nav.values[-1]/nav.values[-5]-1,3)
    if len(nav)>20:
        res['monthRet'] = round(nav.values[-1]/nav.values[-20]-1,3)
    if len(nav)>60:
        res['month3Ret'] = round(nav.values[-1]/nav.values[-60]-1,3)
    if len(nav)>120:
        res['month6Ret'] = round(nav.values[-1]/nav.values[-120]-1,3)
    if len(nav)>250:
        res['yearRet'] = round(nav.values[-1]/nav.values[-250]-1,3)
    return res

tem = DataAPI.FundNavGet(dataDate=u"",secID='',ticker='001410,001938',beginDate='',endDate='',field=u"secID,endDate,ADJUST_NAV",pandas="1").set_index('endDate')
# tem = tem.set_index('endDate')
res = tem.groupby('secID').apply(lambda x:clc_nav(x['ADJUST_NAV']))
```
