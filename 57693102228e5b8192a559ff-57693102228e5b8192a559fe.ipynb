Alpha冲对模型 (多信号合成)

回测框架&基础工作简介：

回测区间从2013年1月1日至今，基准为沪深300，策略每月第一个交易日开盘之后建仓

因子选取：净利润增长率（NetProfitGrowRate）、权益收益率（ROE）、相对强弱指标（RSI）

因子到信号的处理：用到了去极值（winsorize）、中性化（neutralize）、标准化（standardize）处理

组合构建：用到了RDP里的simple_long_only()
