## **1. 本福特定律简介**
本福特定律，也称为第一数定律，是一堆从实际生活得出的数据中观察到以1为首位数字的数出现的机率大约是总数的三成。并且定律说明在许多自然产生的数字中，数字的首位通常是比较小的。本福德早年在通用电器公司（GE）实验室工作，是一名物理学家，二十世纪二十年代发现了一个令人震惊的数学规律，即在任何一组同质随机发生的数据中，排在数据第一或第二位的数字是存有一个可预测到的概率。例如，在一组数据中1排在第一位的概率约为31%，而9排在第一位的概率仅有5%。本福德测试了多种来源的数据组发现存在这样的概率。其数学公式表达为：
$$对于数字的第一位来说P_d=log_{10}(1+1/d)$$ 
$$其中d = {1,2,...,9}$$
$$对于数字的第二位来说P_d=\sum_{k=1}^9log_{10}(1+1/(10k+d))$$
$$其中d = {0,1,2,...,9}$$
$$某个k位数出现的概率P(d_1,d_2,..,d_k)=log_{10}[1+(\sum_{i=1}^kd_i*10^{k-i})^{-1}]$$

依据上面的公式，我编写一个函数计算任意位数字出现的概率（放在附录），如下给出了各位数本福特分布的期望值。

```python
table_one = pd.DataFrame(range(10),index=range(10),columns=['Digit'])
table_one['1st_place'] = table_one['Digit'].map(lambda x:benford_law(x,1))
table_one['2st_place'] = table_one['Digit'].map(lambda x:benford_law(x,2))
table_one['3st_place'] = table_one['Digit'].map(lambda x:benford_law(x,3))
table_one['4st_place'] = table_one['Digit'].map(lambda x:benford_law(x,4))
table_one
```

从上表中可以看出，第一位数字出现的频率有较大差异，而第2，3，4位则趋于相同，画图更能直观看出。

```python
fig = plt.figure(figsize=(12,8))
ax = fig.add_subplot(111)
ax.plot(table_one['Digit'].values,table_one['1st_place'].values,'-bo')
ax.plot(table_one['Digit'].values,table_one['2st_place'].values,'-ro')
ax.plot(table_one['Digit'].values,table_one['3st_place'].values,'-yo')
ax.plot(table_one['Digit'].values,table_one['4st_place'].values,'-co')
def myfunc(x, pos=0): 
    return "{0:.2f}%".format(x * 100)
ax.yaxis.set_major_formatter(FuncFormatter(myfunc))
ax.legend([u"1st_place",u"2st_place",u"3st_place",u"4st_place"],loc='upper right',prop=font,handlelength = 2,handletextpad=0.5,borderpad = 0.1)
ax.set_title(u"Expected Frequencies Based on Benford's Law", fontproperties=font, fontsize=18)
ax.set_ylabel(u"Frequencies", fontproperties=font)
ax.grid(True)
```

从图中更可以看出，第3、4位的曲线基本都相同了，通常我们会拿第一位的本福特定律的来进行分析。

## **3. 本福特定律的理论基础及适用范围**

Hill(1995)给出了本福特定律的一个证明，他指出本福特分布像正态分布一样，是一个经验观察到的现象。他的证明基于这样一个事实，满足本福特分布的样本里面的数是来自其他分布里面的数的组合。如果随机挑选满足其他分布的样本，它们合并样本的各位数将渐进服从本福特分布，尽管它们各自的分布未必服从本福特分布。Hesman(1999)表示关键在于组合来自不同数据源的数字，换句话说，随机合并来自不同分布的不相关的数，它们将服从本福特分布。Boyle(1994)表示来自很多个的数据源的随机变量的数字经过乘除、整数次方后的数据集会服从本福特分布。这可以解释为什么有些会计项目会近似服从本福特分布。一个简单的例子，譬如应收账款、营业收入是由各种商品卖的数理乘以它们对应的价格，或者是营业支出等等。

## **4. 本福特定律在A股财务报表上的实证分析**
-----
依据上面的本福特分析在会计项目上的适用性，我最终决定选用利润表的营业收入(`revenue`)、营业成本(`COGS`)和资产负债表中的应收账款(`AR`)、和应付账款(`AP`)做为我们的数据源，分析最近13年以来所有非金融A股的财务报表。

如下函数可以得到每个股票第一位的频数及样本总数

```python
def get_actual_freq(stk):
    rev_COGS = DataAPI.FdmtISGet(ticker=u"",secID=stk,reportType=u"",endDate=u"",beginDate=u"20130101",publishDateEnd=u"",publishDateBegin=u"",
                      endDateRep="",beginDateRep="",beginYear="",endYear="",fiscalPeriod="",field=u"revenue,COGS",pandas="1")
    AP_AR = DataAPI.FdmtBSGet(ticker=u"",secID=stk,reportType=u"",endDateRep="",beginDateRep=""                ,beginYear="",endYear="",fiscalPeriod="",endDate=u"",beginDate=u"20130101",publishDateEnd=u"",publishDateBegin=u"",field=u"AP,AR",pandas="1")
    def get_digit_one(x):
        return str(x.values[0])[0],str(x.values[1])[0]
    digit_one_income = rev_COGS.apply(lambda x:get_digit_one(x),axis=1)
    digit_one_BS = AP_AR.apply(lambda x:get_digit_one(x),axis=1)
    digit_summary = pd.concat([digit_one_income.map(lambda x:x[0]),digit_one_income.map(lambda x:x[1]),digit_one_BS.map(lambda x:x[0]),digit_one_BS.map(lambda x:x[1])],axis=0).reset_index(drop=True)

    freq = 1.0*digit_summary.value_counts() / len(digit_summary)
    return freq.sort_index(),len(digit_summary)
```

Nigrini M.(1996)提供了z检验来检验各频数是否服从本福特分布，如下所示：
$$z = \frac{|p_o-p_e|-\frac{1}{2n}}{s_i}$$
$$s_i = [\frac{p_e(1-p_e)}{n}]^{1/2}$$
$$其中n是样本总数，\frac{1}{2n}是连续性校正，p_e是本福特分布的期望频率，p_o是样本中对应位数的频率。$$

```python
def z_test(p_zero,p_expect,n):
    
    s = np.sqrt(p_expect*(1-p_expect)/n)
    if abs(p_zero-p_expect) > 1.0/(2*n):
        z_value = (abs(p_zero-p_expect) - 1.0/(2*n))/s
    else:
        z_value = abs(p_zero-p_expect) / s
    return z_value
```

```python
universe = set(set_universe('A')) - set(set_universe(IndSW.YinHangL1)) - set(set_universe(IndSW.FeiYinJinRongL1))
universe = list(universe)
```

```python
benford_dict = {}
for stk in universe:
    freq,length= get_actual_freq(stk)
    benford_dict[stk] = [z_test(freq[0],benford_law(1,1),length),length]
```

```python
result = pd.DataFrame(benford_dict).T.rename(columns={0:'z_test',1:'num'})
```

如上我们得到了每只股票本福特分析第一位值，当`z_test`大于`1.96`时，我们认为在`95%`显著水平应当拒绝原假设样本第一位值的频数与本福特分析的期望频数无差异，也就是说可能存在财务造假。

本福特分布函数

```python
#coding=utf-8

import math
import itertools
def benford_law(digit,level=1):
    if level == 1:
        if digit == 0:
            return None
        else:
            p_sum = math.log10(1+1.0/digit)
            return p_sum
    elif level == 2:
        p_sum = 0
        d1 = range(1,10)
        for i in d1:
            p_sum = p_sum + math.log10(1+1.0/(i*10 + digit))
        return p_sum
    elif level == 3:
        p_sum = 0
        d1 = range(1,10)
        d2 = range(10)
        for item in itertools.product(d1,d2):
            p_sum = p_sum + math.log10(1+1.0/(item[0]*100 + item[1]*10 + digit))
        return p_sum
    else:
        p_sum = 0
        d1 = range(1,10)
        d = range(10)
        for i in d1:
            for item in itertools.product(d,repeat=level-2):
                num_sum = i*math.pow(10,level-1) + sum([item[j]*math.pow(10,level-2-j) for j in range(len(item))]) + digit
                p_sum = p_sum + math.log10(1 + 1.0/num_sum)
        return p_sum
```
