```python
import pandas as pd
import datetime
from CAL.PyCAL import *

#A股股票
allSecList = DataAPI.EquGet(equTypeCD = 'A', field = ['secID'])['secID'].tolist()

#X：2016年6月30日没停牌股票的最高价
X = DataAPI.MktEqudAdjGet(tradeDate=u"20160630", isOpen = "1", field=u'secID,secShortName,tradeDate,highestPrice',pandas="1")
X = X[[i in allSecList for i in X['secID'].values]]
X.index = range(len(X))

#2013年以来历史最高价
maxPrice = []
for sec in X['secID'].values:
    x = DataAPI.MktEqumAdjGet(secID = sec ,beginDate=u"20130101",endDate=u"",field=u"highestPrice",pandas="1")
    maxPrice.append(max(x['highestPrice']))
X['maxPrice'] = maxPrice

#Y：当日最高价大于等于历史最高价
Y = X[X['highestPrice'] >= X['maxPrice']]
Y['listDate'] = DataAPI.EquGet(secID=Y['secID'].tolist(),field=u"listDate",pandas="1")['listDate'].values
Y = Y.sort('listDate')
Y.index = range(len(Y))
print '今天创2013年以来历史新高的股票数量为：', len(Y), '只'
print Y.to_html()
```
