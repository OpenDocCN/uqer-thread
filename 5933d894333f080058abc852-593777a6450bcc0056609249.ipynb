```python
# 注意：数据需要一一对应
# 股票列表：上海机电，华域汽车
list_ticker = ['600835', '600741']
# 可接受的回本年数：6.66,5.66
list_years = [1/0.15 , 1/0.15-1]
# 安全边际:10%，15%
list_marginOfSafety = [0.1, 0.15]
# 退出上浮空间
list_outpct = [0.1, 0.1]
# 现金流增长率：10%，-3%
list_increase = [0.1, -0.03]
# 折现率：8%
discount = 0.08
```

```python
import pandas as pd
import numpy as np

# 企业价值倍数估值
def getEnterpriceValueByROIC(ticker, N = 7, marginOfSafety = 0.2):
    # 平均的ROIC
    df_roic = DataAPI.FdmtIndiRtnGet(ticker=ticker,secID="",endDate="",beginDate="",beginYear=u"",endYear=u"",reportType=u"A",field=u"",pandas="1")

    avg_roic = df_roic['ROIC'].head(5).mean()
    result_roic = df_roic['ROIC'].tolist()[0]

    if result_roic > avg_roic:
        result_roic = avg_roic

    result_roic /= 100
    
    # IC = 股东权益（TShEquity） + 有息负债
    df_IC = DataAPI.FdmtBSAllLatestGet(ticker=ticker,secID=u"",reportType=u"",endDate=u"",beginDate=u"",year="",field=u"",pandas="1")
    # 有息负债 = 短期借款(STBorr) + 1年内到期的长期负债（NCLWithin1Y） + 长期借款（LTBorr） + 应付债券（bondPayable） + 长期应付款（LTPayable）
    df_IC_result = df_IC[['TShEquity','STBorr','NCLWithin1Y','LTBorr','bondPayable','LTPayable']]
    IC = df_IC_result.loc[0].sum()
    
    # EBIT
    EBIT = result_roic * IC

    # EV
    EV = N * EBIT
    
    # 有息负债
    df_IC_debt = df_IC[['STBorr','NCLWithin1Y','LTBorr','bondPayable','LTPayable']]
    debt = np.nan_to_num(df_IC_debt.loc[0].sum())
    
    # 现金
    cash = df_IC.loc[0].cashCEquiv
    
    # 可接受的总市值
    markcap = EV - debt + cash
    
    # 最新报告日期
    reportDateStr =  df_IC.endDate.iloc[0]
    # 当日
    todayStr = pd.datetime.today().strftime('%Y-%m-%d')
    
    # 分红
    df_EDG = DataAPI.EquDivGet(exDivDate="",eventProcessCD=u"",secID=u"",ticker=ticker,beginDate=u"",endDate=u"",field=u"",pandas="1")
    
    # 当前日期大于派息日,最新报告日期小于派息日
    if (todayStr >= df_EDG.iloc[-1].payCashDate) and (reportDateStr < df_EDG.iloc[-1].payCashDate ) :
        totalShares = df_EDG.iloc[-1].sharesAfDiv
        perCashDiv = df_EDG.iloc[-1].perCashDiv
    else:
        totalShares = df_EDG.iloc[-1].sharesBfDiv
        perCashDiv = 0
        

    #  总股本
    if np.isnan(totalShares):
        df_totalShares = DataAPI.EquShareGet(secID=u"",ticker=ticker,beginDate=u"",endDate=u"",partyID=u"",field=u"",pandas="1")
        totalShares = df_totalShares.loc[0].totalShares
        

    # 可接受收盘价
    price = markcap / totalShares 
    
    # 安全边际打折后的收盘价
    price = round(price*(1-marginOfSafety) - perCashDiv,2)
    
    return price

   
# 现金流估值
def getEnterpriceValueByCash(ticker,increase = 0.1, discount = 0.08, marginOfSafety = 0.2, isdataframe = False):
    df = DataAPI.FdmtCFAllLatestGet(ticker=ticker,secID=u"",reportType=u"A",endDate=u"",beginDate=u"",year="",field=u"",pandas="1")
    df['cashflow'] = df['NCFOperateA'] - df['NCFFrInvestA']
    df['preCashflow'] = df['cashflow'] 
    df['preCashflow'][df.index] = df['cashflow'][df.index+1]
    df['pct_chg'] = df['cashflow'] / df['preCashflow'] - 1
    
    df_temp = pd.DataFrame(np.zeros(10),columns = ['净现金流'])
    
    index_increase = "净现金流增长率：%.0f%%" % (increase*100)

    df_temp['净现金流'] = df['cashflow'].tolist()[0]
    df_temp[index_increase] = df_temp.index
    df_temp['净现金流'] = df_temp['净现金流'] * (1 + increase)**df_temp[index_increase]
    
    df_temp['折现后价值'] = df_temp['净现金流']*(1-discount)**df_temp[index_increase]
    
    # 最新报告日期
    reportDateStr =  df.endDate.iloc[0]
    # 当日
    todayStr = pd.datetime.today().strftime('%Y-%m-%d')
    
    # 分红
    df_EDG = DataAPI.EquDivGet(exDivDate="",eventProcessCD=u"",secID=u"",ticker=ticker,beginDate=u"",endDate=u"",field=u"",pandas="1")
    
    # 当前日期大于派息日,最新报告日期小于派息日
    if (todayStr >= df_EDG.iloc[-1].payCashDate) and (reportDateStr < df_EDG.iloc[-1].payCashDate ) :
        totalShares = df_EDG.iloc[-1].sharesAfDiv
        perCashDiv = df_EDG.iloc[-1].perCashDiv
    else:
        totalShares = df_EDG.iloc[-1].sharesBfDiv
        perCashDiv = 0
    
        
    # 可接受的总市值
    markcap = df_temp['折现后价值'].sum()
    
    #  总股本
    if np.isnan(totalShares):
        df_totalShares = DataAPI.EquShareGet(secID=u"",ticker=ticker,beginDate=u"",endDate=u"",partyID=u"",field=u"",pandas="1")
        totalShares = df_totalShares.loc[0].totalShares

    # 可接受收盘价
    price = markcap / totalShares
    
    # 安全边际打折后的收盘价
    price = round(price*(1-marginOfSafety) - perCashDiv,2)
    
    df_temp[index_increase] += 2017
    df_temp[index_increase] = df_temp[index_increase].astype(str) + "年"
    df_temp = df_temp.set_index(index_increase)
    df_temp['净现金流'] = np.round(df_temp['净现金流'] / 100000000,2).astype(str) + "亿"
    df_temp['折现后价值'] = np.round(df_temp['折现后价值'] / 100000000,2).astype(str) + "亿"
    df_temp['可接受的总市值'] = ""
    df_temp['可接受的总市值'].ix[-1] = np.round(markcap / 100000000,2).astype(str) + "亿"
    df_temp['总股本'] = ""
    df_temp['总股本'].ix[-1] = str(round(totalShares / 100000000.0,2)) + "亿股"
    df_temp['安全边际'] = ""
    df_temp['安全边际'].ix[-1] = "%.0f%%" % (marginOfSafety*100)
    df_temp['估值'] = ""
    df_temp['估值'].ix[-1] = str(price) + "元"
    
    if isdataframe:
        return df_temp.T
    else:
        return price

def getClosePrice(ticker):
    price_df = DataAPI.MktEqudGet(tradeDate=u"",secID=u"",ticker=ticker,beginDate=u"",endDate=u"",isOpen="",field=u"",pandas="1")
    
    closePrice = price_df.closePrice.tolist()[-1]
    secID = price_df.secID.tolist()[-1]
    
    try:
        price_df1 = DataAPI.MktBarRTIntraDayGet(securityID=secID,startTime=u"",endTime=u"",unit=u"",pandas="1")
        closePrice = price_df1.closePrice.tolist()[-1]
    except:
        pass
    
    return closePrice


def getStockName(ticker):
    df_name = DataAPI.SecIDGet(partyID=u"",assetClass=u"",ticker=ticker,cnSpell=u"",field=u"",pandas="1")
    name = df_name.secShortName.loc[0]
    return name


# 得到估值数据
def getValueResult(dict_data):
    columns = ['证券名称','当前价格','现价小于估值','估值(已扣除安全边际)','估值','退出价格','现金流估值(已扣除安全边际)']
    df_result = pd.DataFrame(np.zeros(len(list_ticker)*len(columns)).reshape(len(list_ticker),len(columns)),index = list_ticker,columns = columns)
    for i in range(len(list_ticker)):
        ticker = dict_data['list_ticker'][i]
        N = dict_data['list_years'][i]
        marginOfSafety = dict_data['list_marginOfSafety'][i]
        outpct = dict_data['list_outpct'][i]
        increase = dict_data['list_increase'][i]
        discount = dict_data['discount']

        stockname = getStockName(ticker)
        closePrice = getClosePrice(ticker)
        EVPrice = getEnterpriceValueByROIC(ticker, N  , marginOfSafety = 0)
        EVPriceWMOS =  getEnterpriceValueByROIC(ticker, N  , marginOfSafety)
        OutPrice = round(EVPrice * (1 + outpct),2)
        CFPrice = getEnterpriceValueByCash(ticker, increase, discount, marginOfSafety)
        if closePrice < EVPriceWMOS:
            lessThenEVPriceWMOS = '是'
        else:
            lessThenEVPriceWMOS = '否'

        df_result.ix[ticker] = [stockname, closePrice, lessThenEVPriceWMOS, EVPriceWMOS, EVPrice, OutPrice, CFPrice]

    return df_result

# 形成数据字典
dict_data = {'list_ticker':list_ticker,
       'list_years':list_years,
        'list_marginOfSafety':list_marginOfSafety,
        'list_outpct':list_outpct,
        'list_increase':list_increase,
        'discount':discount
       }

# 数据打印
getValueResult(dict_data)
```
