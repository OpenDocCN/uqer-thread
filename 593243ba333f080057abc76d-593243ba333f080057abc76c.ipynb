## 基金评级筛选策略

我们知道，市面上有不少机构提供对公募基金和私募基金的评级业务。

优矿提供了上海证券的基金评级。

![](http://storage-uqer.datayes.com/54d76c61f9f06c276f651a6c/3995af96-4819-11e7-8299-0242ac140004)

其中，评级年限分为三年期和五年期。综合评级分为，1-一星级，2-二星级，3-三星级，4-四星级，5-五星级。

五星级当然是最好的，四星级也不错。

于是我有了一个想法，我在一定时间的调仓频率下，

- 时刻保持最新的四星级的基金，作为我的资产池。

- 当有基金调入调出，我也跟随着进行调仓。
