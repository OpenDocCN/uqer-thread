# 统计风险模型初探

**前言**
--
我们知道，在估计资产之间的协方差矩阵时，基于历史收益数据估计股票的协方差矩阵，经常出现两个问题：
* 估计的参数太多，导致模型不够稳健
* 当资产的个数高于观测的期数之后，计算得到的协方差矩阵不可逆

如何解决这个问题，前有东方证券的压缩矩估计的方法，现有优矿的基本面风险模型。今天再介绍一个新的思路———`统计风险模型`。

**1. 统计风险模型简介**
--
**1.1 基本面风险模型**
既然叫统计风险模型，那么肯定是使用了统计学的方法。我们不妨先看基本面风险模型是怎么进行建模的：
$$r_{n}(t)=\sum_{k}X_{n,k}(t)\*b_{k}{t}+u_{n}(t)$$
或者：
$$r=XF+u$$
在这里面，每个因子都有具体的金融含义。优矿将其分成10个风格因子，28个行业因子和国家因子。但就从式子而言，学过统计的可能会想到什么，没错就是我们统计课上所经常所属的因子分析的数学模型。

**1.2 因子分析模型回顾**
设$r_{i}(i=1,2,...,n)$n个变量，如果表示为：
$$r_{i}=\mu_{i} + a_{i1}F_{1} + ... + a_{ik}F_{1} + \epsilon_{i}$$
我们称$F_{1}, F_{2}, ... , F_{k}$为公共因子，是不可观测的变量，这里就和barra的风险模型有区别啦，它们的系数统计学上叫因子载荷，barra里面叫因子收益率。$\epsilon_{i}$是特殊因子，是不能被前k个公共因子所包含的部分，并且满足：
$$cov(F, \epsilon)=0$$
$$D(F)=I$$
$$D(\epsilon)=
\left[
\begin{matrix}
 \sigma_{1}^{2} \\
 & \sigma_{2}^{2}\\
  &  & \ddots &\\
 &  & & & \sigma_{k}^{2}       \\
\end{matrix}
\right]
$$
这三个式子分别说明了：
* $F,\epsilon$互不相关
* $F_{1}, F_{2}, ... , F_{k}$互不相关，且方差为1
* $\epsilon$互不相关，方差不一定相等，即$\epsilon_{i}$ ~ $N(0, \sigma_{i}^{2})$

**1.3 因子载荷的估计————主成分分析法**
我们暂且不研究其他的性质，将目光放在因子载荷矩阵的估计上。一般来说，我们采用比较多的是主成分分析的方法。假设随机向量$r$的均值为$\mu$，协方差矩阵为$\Sigma$，$\lambda_{1} >= \lambda_{2} >= ...  $为$\Sigma的特征根$，$u_{1}, u_{2}, ...$为对应的标准化特征向量，则：
$$\Sigma=U
\left[
\begin{matrix}
 \sigma_{1}^{2} \\
 & \sigma_{2}^{2}\\
  &  & \ddots &\\
 &  & & & \sigma_{k}^{2}       \\
\end{matrix}
\right]U'=AA'+D$$
具体地：
$$\Sigma=U
\left[
\begin{matrix}
 \sigma_{1}^{2} \\
 & \sigma_{2}^{2}\\
  &  & \ddots &\\
 &  & & & \sigma_{k}^{2}       \\
\end{matrix}
\right]U'=\lambda_{1}u_{1}u_{1}'+...+\lambda_{m}u_{m}u_{m}'+\lambda_{m+1}u_{m+1}u_{m+1}+...+\lambda_{p}u_{p}u_{p}'
         = [\sqrt{\lambda_{1}}u_{1}, \sqrt{\lambda_{2}}u_{2}, ..., \sqrt{\lambda_{p}}u_{p}]\left[
\begin{matrix}
\sqrt{\lambda_{1}}u_{1}' \\
\sqrt{\lambda_{2}}u_{2}'\\
 \vdots \\
 \sqrt{\lambda_{p}}u_{p}'\\
\end{matrix}
\right]$$
上式给出的\Sigma的表达式是精确的，然而，它实际上是没有价值的，因为我们目的是寻求少数几个公共因子去解释，所以会略去后面的p-m项
$$\Sigma\approx \hat{A}\hat{A'}+\hat{D}=\lambda_{1}u_{1}u_{1}'+...+\lambda_{m}u_{m}u_{m}'+\hat{D}
= [\sqrt{\lambda_{1}}u_{1}, \sqrt{\lambda_{2}}u_{2}, ..., \sqrt{\lambda_{m}}u_{m}]\left[
\begin{matrix}
\sqrt{\lambda_{1}}u_{1}' \\
\sqrt{\lambda_{2}}u_{2}'\\
 \vdots \\
 \sqrt{\lambda_{m}}u_{m}'\\
\end{matrix}
\right] + \hat{D}$$
其中$\hat{D}=diag(\hat{\sigma_{1}^{2}}, \hat{\sigma_{2}^{2}}, ..., \hat{\sigma_{p}^{2}})$

**2. Bias Statistics**
--
我们当然要看这个模型是否估计的准确，这里参考了barra的一个做法。我对沪深300的成份股构造了一个等权的组合，然后，算每期的收益（月度）$r_{t}$，t=1, 2, 3, ..., T，同时也计算每一期预测的风险$\sigma_{t}$，t=1, 2, 3, ..., 最后，我们计算标准化的收益$Z_{t}$：
$$Z_{t}=\frac{r_{t}}{\sigma_{t}}$$
偏差统计量定义为$Z_{t}$的标准差：
$$Bias Statistics = \sqrt{var(Z_{t})}$$
如果我们预测的准确的话，那么这个统计了应该很接近于1，如果显著地小于1，则表明我们高估了风险，如果显著的大于1，则我们低估了风险。事实上，在大样本的情况下，假设收益服从正态分布，那么，Bias Statistics的95%置信区间为：
$$[1-\sqrt{2/T}, 1+\sqrt{2/T}]$$
如果我们计算的统计了落在这个区间内，则表明我们需要接受零假设，也就是说我们的预测是准确的。

```python
from lib.stats_riskmdl import *
import pandas as pd
import numpy as np
from CAL.PyCAL import *    # CAL.PyCAL中包含font

# 获取沪深300历史行情
universe = DynamicUniverse('HS300')

start_date = '20070101'
end_date = '20170731'
cal_dates = DataAPI.TradeCalGet(exchangeCD=u"XSHG", beginDate=start_date, endDate=end_date).sort('calendarDate')
cal_dates = cal_dates[cal_dates['isOpen']==1]
trade_date_list = cal_dates['calendarDate'].values.tolist()
cal_dates_ = cal_dates[cal_dates['isMonthEnd']==1]
trade_month_list = cal_dates_['calendarDate'].values.tolist()

univ = []
for date in trade_date_list:
    univ = list(set(univ + universe.preview(date, skip_halted=True)))
ret = DataAPI.MktEqudAdjGet(secID=univ, beginDate=start_date,endDate=end_date, field=u"secID,tradeDate,closePrice",pandas="1")
ret = ret.pivot(index='tradeDate', columns='secID', values='closePrice')
ret = ret.loc[trade_month_list, :]
ret = (ret.pct_change()).iloc[1:, :]
ret = ret.T

sigma = []
r = []
for i in range(len(ret.columns)-13):
	current_univ = universe.preview(ret.columns[i], skip_halted=True)
	current_ret = ret.ix[current_univ, i:i+13]
	current_ret = current_ret.dropna(axis=0, how='any')
	for j in current_ret.index:
		tmp = current_ret.loc[j, :]
		if len(tmp[tmp == 0]) >= 3:
			current_ret.drop(j, inplace=True)
	result = static_riskmdl(current_ret.iloc[:, :12].as_matrix())
	cov_mat = result['cov_mat']
	weights = np.mat([1./ current_ret.shape[0] for i in range(int(current_ret.shape[0]))])
	sigma.append(np.sqrt((weights * cov_mat * weights.T)[0, 0]))
	r.append(weights.dot(np.array(current_ret.iloc[:, -1]))[0, 0])
bt = np.array(r)/np.array(sigma)
print bt.std(ddof=1)
print 1 - np.sqrt(2./len(bt)), 1 + np.sqrt(2./len(bt))
```

这里只是简单的测试，可以看到我们的统计量是在置信区间之内的，所以从这个角度来说，通过主成分方法来估计样本协方差矩阵是可行的。

**3. 统计风险模型的一些思考与不足**
--
先说不足，看起来，我们有了一个计算股票之间协方差矩阵的新方法，但是，别得意，还有些不足。基本面风险模型的功能远不只算协方差矩阵这一点，像业绩归因，如果统计风险模型是不能做到的（想想看为什么），对市场风格的预测统计风险模型也不能做到，这两点也很重要。
再来谈谈一些思考：
* 我们舍弃了后面p-m个主成分，那么这个m又该如何确定，主成分取多少比较合适？
* 当股票数量很大是，计算主成分的会比较耗时间，又该如何解决？
* 对协方差矩阵进行分解，是不是可以考虑对相关系数矩阵进行分解？
* 是不是很多情况下，第一个主成分的功贡献率太高，导致我们的模型会把其他主成分舍去（这个问题和第一个问题一起考虑）？
