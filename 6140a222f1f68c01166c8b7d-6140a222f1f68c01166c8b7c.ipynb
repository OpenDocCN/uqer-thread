# 优选主动基金单因子轮动-估值择时

# 使用动量因子选择主动基金。
## 思路
思路很简单，不用任何复杂操作，明白原理不用写程序在炒股软件上也能实施。
首先，选择所有**主动混合基金**作为标的池。（如果优矿数据没问题，场内主动混合基金只有几十个）
然后，每月月初找出上月涨幅最大的5只基金，等权买入。
最后，每天收盘关注一下买入的基金有没有相对10日前收盘价下跌超过10%的，有就止损买入债券基金。

## 回测
这里没用常用的沪深300做基准，有点欺负人。基准是中证偏股型基金指数。
从2015年1月到2021年9月，基准年化收益率12.7%,策略年化收益率29%，最大回撤20%，

## 利润来源分析
1. 止损。作为动量策略，止损是必须的。如果不止损，必然吃到2015年的股灾，到2018年前基本上没有什么超额收益。这里程序有个bug（我懒得改了），2015年下半年有几次头天买了第二天就止损卖出。所以如果要实盘要注意，如果上月排名前5名的已经从10日前下跌10%就不要买了，也不要用后面的替补，这一份直接空仓。
2. 动量收益。如上所述在2018年前排除止损操作后超额收益不大，主要超额收益来源于市场分化时，比如这几个月。
3. 如果觉得20%的回撤还比较大，附赠估值分位数择时函数，策略程序中28行注释掉用27行，去掉40行的注释就行。加上估值分位数择时收益降到18%，最大回撤降到14%。

```python
#计算当下市场热度，返回分位数
import pandas as pd
from CAL.PyCAL import *
def get_valuation_signal(trade_date):
    #定义指数为沪深300
    code='000300'
    #定义参照时间为前5年
    cal = Calendar('China.SSE')
    dt =datetime.strptime(trade_date,'%Y%m%d')
    end_date = dt.strftime("%Y%m%d")  # 获取回测当日的前1天日期
    start_date = cal.advanceDate(dt,'-5Y',BizDayConvention.Preceding).strftime("%Y%m%d") # 获取回测当日的前21天日期
    #print(start_date,end_date)
    #获得前5年加权PB值
    data=DataAPI.MktIdxdEvalGet(ticker=code,beginDate=start_date,                           endDate=end_date,field=u"PB",pandas="1",PEType=1)
    data=data['PB']
    #print(data)
    pos=0
    if len(data)>0:
        data_now=data.iloc[-1]
        #print(data_now,data.quantile(0),data.quantile(1))
        for f in range(0,10):
            if data_now>=data.quantile(float(f)/10) and data_now<data.quantile(float(f+1)/10):
                pos=(f+(data_now-data.quantile(float(f)/10))/(data.quantile(float(f+1)/10)-data.quantile(float(f)/10)))*10
                #print(pos)
    return pos/100.0
```

```python
# 获取所有场内混合基金
def get_all_stock_fund():
    funds=DataAPI.FundGet(listStatusCd=u"L",category=u"H",operationMode=u"O",field=u"secID",pandas="1")
    return list(funds['secID'])
```

```python
#获取当前存续的混合基金
def get_stock_fund(tradeDate):
    funds=DataAPI.FundGet(listStatusCd=u"L",category=u"H",operationMode=u"O",
                          beginDate=tradeDate,endDate=tradeDate,field=u"secID",pandas="1")
    return list(funds['secID'])
```

```python
stock=get_all_stock_fund()#获取所有混合基金作为universe
```

```python
# 每月计算当月持仓基金
def get_etf(context):
    previous_date=context.previous_date.strftime('%Y-%m-%d')
    funds=get_stock_fund(previous_date)
    funds=list(set(funds).intersection(set(stock)))
    indicator='annTotReturn'
    #获取性能参数
    ind= DataAPI.FundPerfIndicGet(secID=funds,dataDate=previous_date,window=u"1",field="secID,"+indicator,pandas="1")
    ind.set_index('secID',inplace=True)
    ind=ind.sum(axis=1).sort_values(ascending=False)
    to_hold=list(ind.iloc[:5].index)
    return to_hold
```
