在中国：换手率=成交量/流通股本×100%，越高的换手率意味着越高的交易活性。
类似于地震的频率往往预示着地质活动的频繁，那么生成新的高山突起或者峡谷的可能就会变大，所以在实际的交易中，剧烈的换手往往可能意味着潜在可能的剧烈方向变化。

```python
【先取出数据看看他们长啥样】通过DataAPI.MktEquwGet这个API获取股票周行情，包含周开高低收量价、涨跌幅、换手率等，注意这个API的数据结果将只会在每周五收盘后刷新。在这里选择一个周周五的数据来试试看。
```

```python
stock1 = DataAPI.MktEquwGet(secID=u"",ticker=u"",weekEndDate='20170217',isOpen="",beginDate=u"",endDate=u"",field=u"secID,secShortName,openPrice,closeprice,highestPrice,lowestPrice,turnoverRate,avgTurnoverRate,chgPct",pandas="1")
stock1.head(20)
```

【第一部分：如果地质活动很剧烈。。。】事实上我们想想，一个地质活动活跃的地区可能现在已经形成价格的高山或者低谷，那不如简单用价格的变动来寻找那些有可能喷发，但是现在还在蠢蠢欲动的裂口。

```python
stock1=stock1[stock1['chgPct']<0.15].dropna()                                        ###涨跌幅小于15%
stock1=stock1[stock1.chgPct>0.05].dropna()                                           ###涨跌幅大于0.5%
stock1["Stock_amplitude"]=stock1.highestPrice-stock1.lowestPrice                     ##计算周振幅
stock1=stock1[stock1.Stock_amplitude!=0].dropna()                                    #周振幅不等于0，直接排除次新开板情况
stock1=stock1[stock1.turnoverRate>0.15].dropna()                                     #周日均换手率大于15%
stock1.set_index('secID',inplace=True)
stock1.head()
```
