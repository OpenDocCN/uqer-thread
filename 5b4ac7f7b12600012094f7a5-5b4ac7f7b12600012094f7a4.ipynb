# 短向量模型_20180704_博

目的：将量价信息向量化，方便后续使用机器学习的方法进行训练预测。
简单来说就是获得下图所示量价的折线图。一旦转为折线图，就方便下一步将其转化为向量。

<img src='http://chuantu.biz/t6/340/1531057938x-1404817617.png'/>

每一段折现可以表示为[持续天数，区间涨跌幅百分点，成交量]等向量
这样一定时间的量价信息可以化简为，一系列向量序列。

#  第一部分：获取基础数据，以沪深300为例 

```python
import CAL.PyCAL as pycal
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from quartz.api import set_universe

import seaborn as sns
sns.set_style('white')

#参数赋值
start='20150101'
end='20151201'
#样例指数
exampleTicker='000300'
#均线等使用等窗口参数
periodWindow=10

#获取指数行情数据
marketIndexData=DataAPI.MktIdxdGet(indexID=u"",ticker=exampleTicker,tradeDate='',beginDate=start,endDate=end,field=['tradeDate','closeIndex'],pandas="1")
#列tradeDate修改为日期类型并且设置为索引
marketIndexData['tradeDate'] = pd.to_datetime(marketIndexData['tradeDate'])
marketIndexData.set_index("tradeDate", inplace=True)
marketIndexData['closeIndex_mean']=marketIndexData['closeIndex']#.rolling(window=int(periodWindow/2)).mean()
```

## 第二部分：合并价格最高点和最低点
观察上图的特征，可以发现。需要找的折线转折点其实就是极值点，基于这个思路，尝试定位出极值点。

```python
#滑动窗口寻找极致点，使用均线避免过大波动
marketIndexData['closeIndex_mean_max']=marketIndexData['closeIndex_mean'].rolling(window=periodWindow).max()
#今日指数取值=区间内的最高取值说明是一个极值点
marketIndexData['closeIndex_mean_max_bool']=marketIndexData.apply(lambda x:1 if x['closeIndex_mean_max']==x['closeIndex_mean'] else 0,axis=1);
#这部分是为了处理，连续创新高（或新低）的逻辑，持有提示今日（T日，T+1日，T+2日，，）为最高价，其实都位于一个向量上，所以如果T+2日是最高价，将之前几日的最高价标记抹掉
marketIndexData['closeIndex_mean_max_bool_recover']=marketIndexData['closeIndex_mean_max_bool'];
#覆盖昨日的虚假最高点(隐含了未来信息)
for i in xrange(len(marketIndexData),0,-1):
    if(marketIndexData['closeIndex_mean_max_bool_recover'][i-1]==0):
        continue
    #将过去几日的信号用0覆盖
    for j in xrange(1,periodWindow):
        if i-1-j<0:
            break
        marketIndexData['closeIndex_mean_max_bool_recover'][i-1-j]=0

marketIndexData['closeIndex_mean_min']=marketIndexData['closeIndex_mean'].rolling(window=periodWindow).min()
marketIndexData['closeIndex_mean_min_bool']=marketIndexData.apply(lambda x:1 if x['closeIndex_mean_min']==x['closeIndex_mean'] else 0,axis=1);
marketIndexData['closeIndex_mean_min_bool_recover']=marketIndexData['closeIndex_mean_min_bool'];
#覆盖昨日的虚假最高点
for i in xrange(len(marketIndexData),0,-1):
    if(marketIndexData['closeIndex_mean_min_bool_recover'][i-1]==0):
        continue
    #将过去几日的信号用0覆盖
    for j in xrange(1,periodWindow):
        if i-1-j<0:
            break
        marketIndexData['closeIndex_mean_min_bool_recover'][i-1-j]=0
#和上面的最大值区分开，使用-1表示极小值
marketIndexData['closeIndex_mean_min_bool_recover']=-marketIndexData['closeIndex_mean_min_bool_recover']

#信号合并(极大值，极小值点合并为一个，1表示极大值点，-1表极小值)
marketIndexData['closeIndex_mean_merge']=marketIndexData['closeIndex_mean_max_bool_recover']+marketIndexData['closeIndex_mean_min_bool_recover']
marketIndexData['closeIndex_mean_merge_value']=(marketIndexData['closeIndex_mean_merge']*marketIndexData['closeIndex_mean']).apply(lambda x:x if x>0 else -x if x<0 else np.nan)

#箱体图
marketIndexData['closeIndex_mean_merge_value_box_f']=marketIndexData['closeIndex_mean_merge_value'].fillna(method='ffill')
marketIndexData['closeIndex_mean_merge_value_box_b']=marketIndexData['closeIndex_mean_merge_value'].fillna(method='bfill')
#线性插值
marketIndexData['closeIndex_mean_merge_value']=marketIndexData['closeIndex_mean_merge_value'].interpolate(method='linear')#tag01：.fillna(method='ffill')
```

## 第三部分：改进
筛选出“T日”，T日满足是前n日的最高点（最低点）并且是后n日的最高点（最低点）

```python
periodWindow=9
marketIndexData['closeIndex_vector']=marketIndexData['closeIndex']
marketIndexData['closeIndex_min']=0
marketIndexData['closeIndex_max']=0
for i in xrange(periodWindow,len(marketIndexData)-periodWindow):
    #如果indexvoc不是极值就标记为-1，否则保留取值
    max=marketIndexData['closeIndex'][i-periodWindow:i+periodWindow].max()
    min=marketIndexData['closeIndex'][i-periodWindow:i+periodWindow].min()
    if marketIndexData.ix[i,'closeIndex'] == max:
        marketIndexData.ix[i,'closeIndex_max']=1
    else: 
        marketIndexData.ix[i,'closeIndex_max']=0
    if marketIndexData.ix[i,'closeIndex'] == min:
        marketIndexData.ix[i,'closeIndex_min'] = -1
    else: 
        marketIndexData.ix[i,'closeIndex_min']=0
# print marketIndexData.head(20)
marketIndexData['closeIndex_merge']=(marketIndexData['closeIndex_max']+marketIndexData['closeIndex_min']).apply(lambda x:np.nan if x==0 else x)
marketIndexData['closeIndex_vector']=(marketIndexData['closeIndex_merge'].apply(lambda x:x if x>0 else -x))*marketIndexData['closeIndex']
marketIndexData['closeIndex_vector_f']= marketIndexData['closeIndex_vector'].fillna(method='ffill')
marketIndexData['closeIndex_vector_b']= marketIndexData['closeIndex_vector'].fillna(method='bfill')
marketIndexData['closeIndex_vector_i']=marketIndexData['closeIndex_vector'].interpolate(method='linear')
# print marketIndexData.head(20)
fig = plt.figure(figsize=(15,6))
ax1 = fig.add_subplot(1,1,1)
ax1.plot(marketIndexData[['closeIndex','closeIndex_vector_f','closeIndex_vector_b','closeIndex_vector_i']])
```

## 第四部分：输出向量格式信息

```python
#注意填充方式为bfill
marketIndexData['closeIndex_merge_fillna']=marketIndexData['closeIndex_merge'].fillna(method='bfill')
marketIndexData['closeIndex_merge_fillna']=marketIndexData['closeIndex_merge_fillna'].fillna(0)
def transToVector(marketData,signalColumn='closeIndex_merge_fillna',closePriceColumn='closeIndex'):
    retDf=pd.DataFrame(columns=['value','count','rate100'])
    if(len(marketData)==0):
        return retDf
    tmpValue=marketData[signalColumn][0]
    tmpCount=1
    startPrice=marketData[closePriceColumn][0]
    endPrice=marketData[closePriceColumn][0]
    for i in range(1,len(marketData)):
        if(marketData[signalColumn][i]!=tmpValue):
            retDf=retDf.append({'value':tmpValue,'count':tmpCount,'rate100':float('%.2f' % (100.0*(endPrice/startPrice-1)))},ignore_index=True)
            tmpValue=marketData[signalColumn][i]
            tmpCount=1
            startPrice=marketData[closePriceColumn][i]
            endPrice=marketData[closePriceColumn][i]
        else:
            tmpCount+=1
            endPrice=marketData[closePriceColumn][i]
    retDf=retDf.append({'value':tmpValue,'count':tmpCount,'rate100':float('%.2f' % (100.0*(endPrice/startPrice-1)))},ignore_index=True)
    return retDf

vectors=transToVector(marketIndexData,'closeIndex_merge_fillna','closeIndex')  
print  (vectors)
```

说明，图中每一行，表示一个单独的向量，和上图的每一条折线对应
比如，第1行（0），value=-1表示跌，count=25表示25天，rate100=9.04表示涨跌幅为-9.04%
需注意的是：最后一行的0表示由.fillna(method='bfill')方式引入的nan，fillna(0)之后得到的
实际含义是尚未计算出的结果（由于此方式依赖未来信息）
从折线图也可以看出，区间最后几日是没有折现的。
