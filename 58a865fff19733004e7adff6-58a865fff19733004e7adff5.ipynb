```python
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
```

这段就是“找个地方画画”的意思，也就是在画板里加一子图，位置：2X2中的第一个位置。下面有一种更简便的方法：

```python
fig,axes=plt.subplots(2,3,sharex=True,sharey=False)#sharex表示所有x坐标相同
```

axes是一个数组，axes[0,1]代表第二个子图，拿个例子比较直观点

```python
fig,axes=plt.subplots(2,2,sharex=True,sharey=True)
for i in range(2):
    for j in range(2):
        axes[i,j].hist(randn(500),bins=50,color='k',alpha=0.5)
plt.subplots_adjust(wspace=0,hspace=0)
```

名词解释：1）sharex/sharey 表示画板横/纵坐标相同；2）hist函数为柱形图，里面bins表示柱形图宽度，color颜色，alhpa分布特点；3）subplots_adjust调整子图与子图间宽度，包括wspace宽度，hspace长度，单位是百分比；

- 要如何才能画好一张图呢？这里有些画图的基本功能可以满足简单的绘图需求。

```python
fig,ax=plt.subplots(1,1)
ax.plot(randn(1000).cumsum(),color='g',linestyle='--')#cumsum()是累加的意思，这样线就连续了
```

.plot中2个常用参数：1）color颜色；2）linestyle线型

```python
fig,ax=plt.subplots(1,1)
ax.set_yticks(range(10,110,10))
ax.set_xticklabels(range(0,11,1))
ax.set_xlabel('xlabel')
ax.set_title('title')
ax.plot(randn(1000).cumsum(),color='g',linestyle='--')
```

子图中的4个参数：1）set_xticks设置x轴的度量间距，这里用y轴设置;2)set_xticklabels添加x轴的标签;3)set_xlabel给x轴取个名字;4)set_title取子图的名字;

- 如何在一张图上画多根线？使用图例的功能，英语叫做legend

```python
fig,ax=plt.subplots(1,1)
ax.plot(randn(1000).cumsum(),color='g',linestyle='--',label='one')
ax.plot(randn(1000).cumsum(),color='r',linestyle='-',label='two')
ax.plot(randn(1000).cumsum(),color='b',label='three')
ax.legend(loc='best')#loc表示位置
```

制作两部曲：1）赋值并加label值；2）用legend合起来；

- 想加个注释标签好做吗？用以下函数方便涂鸦

```python
fig,ax=plt.subplots(1,1)
ax.plot(randn(1000).cumsum(),color='g',linestyle='--',)
ax.text(200,0,'label1',family='monospace',fontsize=12)#不加这个family会和子图内容重叠
```
