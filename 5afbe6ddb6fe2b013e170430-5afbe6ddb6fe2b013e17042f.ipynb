二. 计算贵州茅台与恒瑞医药2018年2月所有交易日的TTM营业总收入和TTM管理费用

```python
#作业中发现的问题
#a.注意审题：营业总收入（包含利息收入、保费收入等），而不是营业收入
#b.单季度和今年以来累计的数据均会被存在数据表里，要选择正确的数据
df = DataAPI.FdmtISGet(ticker=u"600519",beginDate=u"20170101",endDate=u"20180228",pandas="1",field="ticker,publishDate,endDate,fiscalPeriod,tRevenue,adminExp")
print(df)
print('================================')
df['endDate'] = pd.to_datetime(df['endDate'])
df = df.loc[df['endDate'].dt.month == df['fiscalPeriod'].astype(int)]
print(df)
```

```python
import pandas as pd
import numpy as np

def get_financial_data(ticker_symbol, read_start_date='20110101'):
    #取财务数据
    df = DataAPI.FdmtISInduGet(ticker=ticker_symbol,beginDateRep=\
                           read_start_date,field=u"ticker,publishDate,endDate,fiscalPeriod,tRevenue,adminExp",pandas="1")
    df['publishDate'] = pd.to_datetime(df['publishDate'])
    df['endDate'] = pd.to_datetime(df['endDate'])
    df = df.loc[df['endDate'].dt.month == df['fiscalPeriod'].astype(int)]
    return df

def get_statement_date(latest_date):
    #获取去年及去年同期财报的endDate
    last_year_end_date = pd.to_datetime(str(latest_date.year - 1) + '1231')
    last_same_season_end_date = pd.to_datetime(int(latest_date.strftime('%Y%m%d')) - 10000, format='%Y%m%d')
    return last_year_end_date, last_same_season_end_date
    
def get_ttm(df, T, col_name):
    #根据历史财报数据df，截止日期T，财务指标名（即净利润还是营收等），构造截至T日的TTM财务指标
    df = df.loc[df['publishDate'] <= T].copy()
    df = df.sort_values(['endDate', 'publishDate'], ascending=[False, False])
    latest_date = df.iloc[0].loc['endDate']
    last_year_end_date, last_same_season_end_date = get_statement_date(latest_date)
    value_this_season = df.loc[df['endDate'] == latest_date]
    if len(df) > 0:
        value_this_season = value_this_season.iloc[0].loc[col_name]
    else:
        value_this_season = np.nan
    value_last_year = df.loc[df['endDate'] == last_year_end_date]
    if len(value_last_year) > 0:
        value_last_year = value_last_year.iloc[0].loc[col_name]
    else:
        value_last_year = np.nan
    value_last_same_season = df.loc[df['endDate'] == last_same_season_end_date]
    if len(value_last_same_season) > 0:
        value_last_same_season = value_last_same_season.iloc[0].loc[col_name]
    else:
        value_last_same_season = np.nan
    value_TTM = value_this_season + value_last_year - value_last_same_season
    return value_TTM

def get_tRevenue_adminExp():
    #获取交易日期
    trade_cal = DataAPI.TradeCalGet(exchangeCD=u"XSHG",beginDate=u"",endDate=u"",field=u"isOpen,calendarDate",pandas="1")
    trade_cal = trade_cal[trade_cal['isOpen']==1]
    trade_cal['calendarDate'] = pd.to_datetime(trade_cal['calendarDate'])
    trade_cal = trade_cal.loc[trade_cal['calendarDate'].between(pd.to_datetime('20180201'), pd.to_datetime('20180228'))]
    #获取股票列表
    tickera_list = ['600519', '600276']
    data_df = pd.DataFrame(columns = ['ticker', 'tradeDate', 'tRevenue', 'adminExp'])
    #构造ACCA因子
    for ticker_symbol in tickera_list:
        #print(ticker_symbol)
        df = get_financial_data(ticker_symbol, read_start_date='20110101')
        for trade_date in trade_cal['calendarDate']:
            tRevenue = get_ttm(df.copy(), trade_date, 'tRevenue')
            adminExp = get_ttm(df.copy(), trade_date, 'adminExp')
            data_df.loc[len(data_df), :] = [ticker_symbol, trade_date, tRevenue, adminExp]
    data_df = data_df.sort_values(['tradeDate', 'ticker'])
    return data_df

data_df = get_tRevenue_adminExp()
```

三.计算2018年2月份所有交易日的贵州茅台与恒瑞医药的Administrative expense rate因子值，并与优矿的比较。
备注：Administrative expense rate（管理费用与营业总收入之比）= 管理费用（TTM）/ 营业总收入（TTM）。

```python
#作业中发现的问题
#a.ttm数据获取没有选择正确的数据, 记住同一天取最大的endDate
#b.ttm取完后忘记pit
ttm_data_uqer=DataAPI.FdmtISInduTTMPITGet(ticker=u"600519,600276",endDate=u"",publishDateBegin=u"20170101",publishDateEnd=u"20180228",field=u"ticker,publishDate,endDate,tRevenue,adminExp",pandas="1")
ttm_data_uqer = ttm_data_uqer.sort_values(['ticker', 'publishDate', 'endDate'])
print(ttm_data_uqer)
```

```python
#正确的取法
ttm_data_uqer=DataAPI.FdmtISInduTTMPITGet(ticker=u"600519,600276",endDate=u"",publishDateBegin=u"20170101",publishDateEnd=u"20180228",field=u"ticker,publishDate,endDate,tRevenue,adminExp",pandas="1")
ttm_data_uqer = ttm_data_uqer.sort_values(['ticker', 'publishDate', 'endDate'])
ttm_data_uqer = ttm_data_uqer.drop_duplicates(['ticker', 'publishDate'], keep='last')
ttm_data_uqer = ttm_data_uqer.rename(columns={'publishDate':'tradeDate'})
ttm_data_uqer = ttm_data_uqer[['ticker', 'tradeDate', 'tRevenue', 'adminExp']]
ttm_data_uqer['tradeDate'] = pd.to_datetime(ttm_data_uqer['tradeDate'])

#获取交易日历
trade_cal = DataAPI.TradeCalGet(exchangeCD=u"XSHG",beginDate=u"",endDate=u"",field=u"isOpen,calendarDate",pandas="1")
trade_cal = trade_cal[trade_cal['isOpen']==1]
trade_cal['calendarDate'] = pd.to_datetime(trade_cal['calendarDate'])
trade_cal = trade_cal.loc[trade_cal['calendarDate'].between(pd.to_datetime('20170101'), pd.to_datetime('20180228'))]
trade_cal = trade_cal.rename(columns={'calendarDate':'tradeDate'}) 

#补充交易日期
ttm_data_uqer = ttm_data_uqer.set_index(['tradeDate', 'ticker']).unstack()
for date in trade_cal['tradeDate']:
    if date not in ttm_data_uqer.index:
        ttm_data_uqer.loc[date, :] = np.nan

#fillna，计算因子
ttm_data_uqer = ttm_data_uqer.sort_index().fillna(method='ffill')
ttm_data_uqer = ttm_data_uqer.stack().reset_index()
ttm_data_uqer['AdminiExpenseRate'] = ttm_data_uqer['adminExp'] / ttm_data_uqer['tRevenue']

ttm_data_uqer.tail()
```

```python
#unstack方法与groupby方法比较

import time
#获取行情数据
df = DataAPI.MktEqudAdjGet(secID=u"",ticker=u"",beginDate=u"20180101",endDate=u"20180509",field=u"ticker,tradeDate,closePrice",pandas="1")
df = df.sort_values(['ticker', 'tradeDate'])
print(df.head())
df_groupby = df.copy()

print('===================')
#用unstack方法计算
t1 = time.time()
df = df.set_index(['tradeDate', 'ticker']).unstack().pct_change().stack(dropna=False).reset_index()
df = df.sort_values(['ticker', 'tradeDate'])
print(df.head())
t2 = time.time()
print('unstack: ', t2 - t1)

print('=====================')
#用groupby方法计算
t3 = time.time()
df_groupby['closePrice'] = df_groupby.groupby('ticker')['closePrice'].pct_change()
df_groupby = df_groupby.sort_values(['ticker', 'tradeDate'])
print(df_groupby.head())
t4 = time.time()
print('groupby: ', t4 - t3)
```
