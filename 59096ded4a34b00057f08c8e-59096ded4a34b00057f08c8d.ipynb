```python
#对比每年收益
year_return={}
start_year=int(start_date[:4])
end_year=int(end_date[:4])
for year in range(start_year,end_year+1):
    year_return[year]=df_result[str(year)].ix[-1]/df_result[str(year)].ix[0]
df_year_return=pd.DataFrame(year_return)
df_year_return.T.plot(kind='bar',figsize=[15,10])
df_year_return
```

```python
#相对基准（行业指数）收益曲线
df_exceed=pd.DataFrame()
for c in df_result.columns:
    df_exceed[c]=df_result[c]-df_result['closeIndex_x']
df_exceed.plot(figsize=[16,10])
```
