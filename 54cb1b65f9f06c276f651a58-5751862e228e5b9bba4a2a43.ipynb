# RDP_TO_PMS

该代码实现的功能：将RDP生成的权重文档进行处理，适用PMS，可以当做导入持仓\_全更新模板

1，需要准备的文档，由RDP生成的权重文档，代码里用的是：'20130131.csv','20130228.csv','20130329.csv'这三个文件；

2，生成文档：PMS.csv，将这个文档导入到PMS中，即可在组合中导入该持仓。

```python

fn_list = ['20130131.csv','20130228.csv','20130329.csv']    #这里是所要处理的RDP过来的文件名
final_df = pd.read_csv('importWeight.sample.new.csv',encoding='GBK')[0:3]
final = pd.DataFrame({})

for fn in fn_list:
    year = fn[0:4] 
    month = fn[4:6] if fn[4]=='0' else fn[5]
    day = fn[6:8] if fn[6]=='0' else fn[7]
    date_pms = '/'.join([year,month,day])
    field = ['openPrice','highestPrice','lowestPrice','closePrice']
    
    lines = pd.read_csv(fn,header=None)
    lines[0] = lines[0].astype(str)
    lines[0] = lines[0].map(lambda x:'0'*(6-len(x))+x)

    #获得当日行情数据，判断该个股是否有停盘、涨跌停现象
    num = 50
    count = len(lines)/num    
    if count>0:
        mkt_info = pd.DataFrame({})
        for i in range(count):
            sub_info = DataAPI.MktEqudGet(ticker=list(lines[0])[i*num:(i+1)*num],beginDate=fn[0:8],endDate=fn[0:8],field=field)
            mkt_info = pd.concat([mkt_info,sub_info])
        sub_info = DataAPI.MktEqudGet(ticker=list(lines[0])[(i+1)*num:],beginDate=fn[0:8],endDate=fn[0:8],field=field)
        mkt_info = pd.concat([mkt_info,sub_info])
    else:
        mkt_info = DataAPI.MktEqudGet(ticker=list(lines[0]),beginDate=fn[0:8],endDate=fn[0:8],field=field) 

    mkt_info = mkt_info.reset_index().drop('index',axis=1)
    #print u'停盘：',mkt_info['secID'][mkt_info['openPrice']==0]
    #print u'涨跌停：',mkt_info['secID'][(mkt_info[field].std(axis=1)==0)&(mkt_info['openPrice']!=0)]
    
    mkt_info[u'日期'] = date_pms
    mkt_info[u'证券代码'] = mkt_info['secID'].map(lambda x:x[0:6])
    mkt_info[u'交易市场代码'] = mkt_info['secID'].map(lambda x:'XSHG' if x[0]=='6' else 'XSHE')    #有一列记录交易市场类型 
    mkt_info[u'持仓数量'] = lines[1].map(lambda x:x*0.98)
    
    sub_df = mkt_info[[u'日期',u'证券代码',u'交易市场代码',u'持仓数量','openPrice']]
    sub_df.rename(columns={'openPrice':u'成本价格'},inplace=True)
    final = pd.concat([final,sub_df])
    

    
final[u'持仓类型'] = 'LONG'
final[u'数量模式'] = 'WEIGHT'

final_df = pd.concat([final_df,final])
final_df = final_df.loc[:,[u'日期',u'证券代码',u'交易市场代码',u'持仓类型',u'持仓数量',u'数量模式',u'成本价格',u'投资类型']]
final_df.to_csv('test.csv',index=False,encoding='GBK')    #生成多日更新持仓文档，可以上传到PMS

            
 
```
