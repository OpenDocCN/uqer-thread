```python
def handle_data(account):                  # 每个交易日的买入卖出指令
    hist_price=pd.DataFrame(account.get_attribute_history('closePrice',1)).T
    hist_price.sort(columns=0,inplace=True,ascending=True)
    newUniverse=hist_price.index.values
    for stock in account.avail_secpos:
        if (stock in account.referencePrice) and (account.referencePrice[stock]>account.price_flag[stock]*(1+sell_return)):
            order_to(stock,0)
            account.piece-=1
    for stock in newUniverse: 
        if account.piece<position_piece:
            if (stock in account.universe) and (not stock in account.valid_secpos):
                try:
                    order_pct_to(stock,1/position_piece)
                except:
                    continue
                account.piece+=1
                account.price_flag[stock]=account.referencePrice[stock]
    return
```
