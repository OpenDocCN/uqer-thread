## 2、官方规则详细解读
---

- 奖励的部分直接跳过，我们现在只认真研读规则

- 可以看到排名有两个组，`股票多头组`和`市场中信组`，因为其风险收益不一样，所以单独排名更公平，另外也可以看到两者的排名规则和操作要求也有差异

- 重点看**策略标准与排名**，下面就大的点作如下说明



### 2.1 策略统一要求
---

- 这一点是对两类策略都共有的要求

- 主要目的是告诉你只能买哪些股票，应该来说Universe还是非常够用的

- 提醒一下，基金不能买哟，另外有买分级B小想法的别继续做梦了，违规可是要付出代价的



### 2.2 模拟交易表现
---

- 重点说明一下，`本次排名不考虑回测结果，完全按照三个月模拟交易表现来（防止回测结果过度拟合、使用未来数据等），但最后答辩路演阶段会重点参考策略回测结果和模拟结果的一致性`，所以请大伙对自己负责，这样更公平些

- 所以这里的**`模拟盘要求表现是硬性的，模拟盘最后一天能排上名的肯定都是满足表现要求的`**（但对于最大回撤指标，只要超过了标准就不再参与排名；而对于年化收益、换手率这种不到最后一天不知道结果的都会继续排名，直到排名最后一天对这个做强制限制），下面对两类要求做详细解释

----

- #### 股票多头组

  - 年化收益率大于30%：要求策略必须是赚钱的，不管市场是牛市还是熊市，所以对策略的操作几乎没有任何限制，甚至可以日内择时
  
  - 最大回撤低于10%：对策略风险的控制，过大的回撤会让投资者望而却步，这点大伙一定要把握住，博收益的同时一定要提前预算好最大风险程度
  
  - 双边年化换手率大于24倍：防止僵尸策略，另外，更多独立判断的次数更能看出投资经理水平
  
  - 单只股票按买入成本计算持股比例不得高于总资产15%：强调组合投资，同时也防止因运气好买到牛逼个股的极端情况（比如之前的万科A）
  
  
- #### 市场中性组

  - 年化收益率大于10%：需要注意这里是指`股票多头+期货多头整体要求的年化收益率`，把这300万当成一个产品就对了
  
  - 最大回撤低于4%：市场中性的特点就是稳健的赚钱，所以风险要求也严格一些
  
  - 换手率+单只个股持仓比：同股票多头
  
  - 股票多头资产低于100万的累计天数不超过3天：理性的市场中性策略都会投入全部的资金，股票多头初始资金200万，这里设置100万就是让大家踏实做alpha，不要想着碰运气的择时
  
  - （0.7,1.3）：理性的市场中性也是全对冲，这里的区间限制就是防止大伙疯狂择时从而导致排名不公平；当然，在风险可控的情况下的择时确实能提高收益的话，强烈支持！
  
  - 期货合约累计开平仓手数不得超过10手（含）：若以当月合约对冲，那么每个自然月就换仓一次，每次换仓开平仓手数4手，这里不超过10手就是防止组合收益都是来自于期货的单边交易，这样或许有运气成分在，同时不公平也不是市场中性的特征
  

### 2.3 排名方式
---

- 同500万众包版对冲基金，先按单指标排名，然后按一定的权重汇总各项的得分计算总得分

- 可以看到，规则中`对收益率的权重比较高（和众包基金不一样）`，而且从之前介绍的模拟盘表现要求来看，对收益的要求都是非常高的，所以重点大家懂的，2000万是一个FOF产品，现在大伙看到基金产品第一个看的是啥指标？好了，点到为止

- 需要提醒的是，追求收益的同时别忘了最大回撤的要求，慎重慎重，请在适当的时候使用你的洪荒之力！


## 3、实操分享
----

- 首先介绍最重要的功能，也就是相比众包基金的亮点

- **`可以替换策略了`**：若在比赛期间想换策略或者发现代码bug等问题，都可以替换策略了，关于替换策略的细节后面有实例进行说明
  
- **`可以手动下单了`**：之前是开盘前产生买卖信号，在上午9:30~9:45择机成交，对这个说再见了！你完全可以根据自己的喜好进行时点的选择进行买卖，只需要在提交策略时选择“下单后需手动确认成交”即可，后续也会有实例说明

- 另外，关于参赛流程就不多说，详细请参见[`官方介绍`](https://uqer.datayes.com/mom/guide)

- 下面就三个方面的实盘经验来分享


### 3.1 替换策略
---

- 新的策略，点击开始交易，选择替换策略，选择好被替换的策略（这个策略的源码就被替换掉了）

- 然后选择相应的参数就好，截图如下，对其中的重要参数做重点说明


![](http://i2.piimg.com/567571/badc28b63dcf303e.jpg)

### `开始运行时间`
---

- 分位当天继续运行handle_data和下个交易日继续运行

- 这一点很好理解，若想马上根据新的策略跑出结果就要选择今天继续运行handle_data，若今天不想变动仓位，则可以选择明天再运行


### `策略参数`
---

- 这一点`非常重要`，请认真对待

- 替换策略简单来讲就是重新定义你的策略代码，但是account下的东西都会继承下来，包括账户里有多少钱、账户持仓等相关信息；这一点和实际情况也最接近，假想account就是你的券商账户，任何时刻你账户里的相关信息都必须保持一致性；下面，对其中的一些细节做更详细说明

  - 新的universe：假想你替换的策略和你之前策略的universe不一致，那么account下的信息就是你新策略的universe与你之前持仓股票的universe取并集
  
  - account.referencePrice等：都跟随新的universe动态更新
  
  - 另外，由于提交策略后是重新跑整个策略，所以若在你的code里有用到自定义package、全局变量的，都已最新运行的信息为准；当然想继续保留initialize下面的信息的参见下一条

- 对于一些喜欢在initialize里自定义各类参数的小伙伴，若提交新策略之后想继续保留这些参数状态，可以选择“保留自定义参数状态”，若想重新初始化initialize里这些参数，请选择根据initialize函数初始化所有自定义参数

### 3.2 手动下单
---

- 若在提交策略时选择了“下单后需手动确认成交”就有此功能，这样下单过程需自己完成，好处是自己可以选择日内任何时点下单，和自己炒股完全一样

- 进入我的交易，点击参赛策略就可以看到在“股票调仓记录”上方有【去确认下单】按钮，点击即进入下单界面，如下图


![](http://i1.piimg.com/567571/266b1a35e9f0505b.jpg)

### 对照界面说明几点：
---

- 点击篮子下面的组合即可得到右边的买卖列表，同时，在这里大伙还可以`对买卖列表做修改调整`，比如我觉得万科A现在风险太高，虽然模型跑不出来，但在万科大战中我不想冒这个风险，于是我将其删除

- 待调整好之后，点击左下角的【下达篮子】按钮即完成手动下单

- 随后，大伙就可以去通联组合管理系统查看自己的收益、持仓等信息（PS：通联组合管理是面向机构的投资分析交易管理工具，价值不菲，机会难得大伙珍惜）


### 3.3 期货对冲
---

- 如上的介绍已经满足了股票多头组的需求，但对于市场中性组的还有期货对冲的部分需要自己去完成

- 进入市场中信策略的【策略配置界面】，点击“去组合管理手动下单”按钮，即可进入组合管理系统进行期货手动下单，如下图所示

### 4、玩转大赛
---

- 这一部分写出来比较长，所以仅仅列出操作思路，供大家参考

- 日内T+0
 - 首先从持仓中确定要卖的股票代码和数量
 - 新建一个简单的策略，universe就为你要卖的股票集合，handle_data()部分就写卖出逻辑，选择替换策略（需要手动确认下单），然后手动选择时点进行卖出
 - 同样的方式，替换策略，买回来，实现T+0

- 日内择时：
 - 大部分原理同日内T+0，只需手动下单确认，自己日内择时成交
