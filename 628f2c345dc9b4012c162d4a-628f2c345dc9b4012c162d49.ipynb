# 分析师预期

```python
import pandas as pd
import numpy
w1 = pd.read_csv('w1.csv', index_col = 0)
w1.index = pd.to_datetime(w1.index)
```

```python
new_c = []
for i in range(len(w1.columns)):
    n = w1.columns[i]
    if n[-1] == 'Z':
        new_c.append(n[:-3]+'.XSHE')
        continue
    if n[-1] == 'H':
        new_c.append(n[:-3]+'.XSHG')
        continue
    new_c.append(n)
w1.columns = new_c
```
