```python
#Spearman 和Pearson相关系数在算法上完全相同. 只是PEARSON相关系数是用原来的数值计算积差相关系数, 而SPEARMAN是用原来数值的秩次计算积差相关系数
# 1.连续数据，正态分布，线性关系，用pearson相关系数是最恰当，当然用spearman相关系数也可以，就是效率没有pearson相关系数高。
# 2.上述任一条件不满足，就用spearman相关系数，不能用pearson相关系数。
# 3.两个定序测量数据之间也用spearman相关系数，不能用pearson相关系数
# Pearson相关系数受异常值影响大

import scipy.stats as st

# 1、构造ROE数据和return数据
np.random.seed(1)
# roe_data = np.random.rand(3000)
# roe_data = [x-0.5 for x in roe_data]
roe_data = np.random.uniform(-0.5, 0.5,3000)#random.uniform生成一个指定范围内的随机符点数

# return_data = np.random.randint(-100, 100, 3000)
# return_data = [1.0*x/10 for x in return_data]
return_data = np.random.uniform(-10, 10,3000)

factor_data = pd.DataFrame({"ROE":roe_data, "return":return_data})
print factor_data.head().to_html()

# 2、将数值转化为排序值
factor_data['ROE_rank'] = factor_data['ROE'].rank()
factor_data['return_rank'] = factor_data['return'].rank()

# 3、根据公式计算相关系数
#计算每列数据均值
mean_v = factor_data.mean()
mean_v
# 计算各个值和均值的差
factor_data['ROE_rank_delta'] = factor_data['ROE_rank']-mean_v['ROE_rank']
factor_data['return_rank_delta'] = factor_data['return_rank']-mean_v['return_rank']
# 将偏差相乘，并累计求和,再除以总个数，得到分子
factor_data['delta_multip'] = factor_data['ROE_rank_delta']*factor_data['return_rank_delta']
up_value = factor_data['delta_multip'].sum()/len(factor_data)
print factor_data.head().to_html()
# 分母是两个的标准差相乘
down_value = factor_data['ROE_rank'].std()*factor_data['return_rank'].std()

# 4、分子/分母
v = up_value/down_value

print '根据公式计算结果：',v
print 'scipy的包计算结果：',st.spearmanr(factor_data['ROE'], factor_data['return'])[0]
print 'spearman相关系数:',factor_data[['ROE_rank','return_rank']].corr('spearman').to_html()
print 'pearson相关系数:',factor_data[['ROE_rank','return_rank']].corr('pearson').to_html()
```

```python
#直观感受spearman和pearson相关系数
import scipy.stats as st
import numpy as np
ROE = np.arange(0,10,2)
Return = np.arange(1,10,2)
df = pd.DataFrame([ROE,Return],index=['ROE','Return']).T
# 0、构造简单数据，观察spearman相关系数和pearson相关系数
df1 = df.copy()
df['ROE_rank'] = df['ROE'].rank()
df['Return_rank'] = df['Return'].rank()
print df
print st.spearmanr(df['ROE'], df['Return'])
print st.pearsonr(df['ROE'], df['Return'])
print '-'*50
# 1、改变数字的大小，但不改变排列顺序，spearman相关系数没有变化，而pearson相关系数有变化
df2 = df1.copy()
df1.loc[2][0] = 5
df1['ROE_rank'] = df1['ROE'].rank()
df1['Return_rank'] = df1['Return'].rank()
print df1
print st.spearmanr(df1['ROE'], df1['Return'])
print st.pearsonr(df1['ROE'], df1['Return'])
print '-'*50
# 2、改变数字的大小，也改变排列顺序，spearman相关系数和pearson相关系数都有变化,pearson相关系数受异常值影响大
df2.loc[1][0] = 20
df2['ROE_rank'] = df2['ROE'].rank()
df2['Return_rank'] = df2['Return'].rank()
print df2
print st.spearmanr(df2['ROE'], df2['Return'])
print st.pearsonr(df2['ROE'], df2['Return'])
```
