### 笔者在此对常用的seaborn作图进行示范，同时也结合了DataAPI的使用以及Pandas的简单操作，都已经在程序中注释，方便初学者使用

### 更多内容，请详见seaborn官方教程

http://web.stanford.edu/~mwaskom/software/seaborn/tutorial.html

```python
#导入常用的库
from CAL.PyCAL import *
import numpy as np
import pandas as pd
from pandas import DataFrame, Series
import matplotlib.pylab as plt
import seaborn as sns

#以下设定数据时间，默认为2014年整年
start_date = DateTime(2013,12,31)
end_date = DateTime(2014,12,31)
start_date = start_date.strftime('%Y%m%d')
end_date = end_date.strftime('%Y%m%d')
#print start_date,end_date

#以银行板块的浦发银行，工商银行，中国银行为例
stock1 = '600000.XSHG'
stock2 = '601398.XSHG'
stock3 = '601988.XSHG'

#获取每只股票的收盘价序列, 这里使用到了DataAPI，可参考笔者的短文 “［量化基础］如何使用DataAPI”
data1 = DataAPI.MktEqudGet(secID = stock1, beginDate = start_date, endDate = end_date, field = ['TradeDate', 'closePrice'])
data2 = DataAPI.MktEqudGet(secID = stock2, beginDate = start_date, endDate = end_date, field = ['TradeDate', 'closePrice'])
data3 = DataAPI.MktEqudGet(secID = stock3, beginDate = start_date, endDate = end_date, field = ['TradeDate', 'closePrice'])

#按照tradeDate列名来合并
price = pd.merge(data1, data2, on = 'tradeDate')
price = pd.merge(price, data3, on = 'tradeDate')

#修改列名
price.columns = ['tradeDate', 'Close1', 'Close2', 'Close3']

#修改索引
price.set_index('tradeDate', inplace = True)
#展示效果如下
price.head()
```

    获得了几只股票的收盘价序列后
    
    利用DataFrame的pct_change()功能，计算收益率；重置索引；并剔除第一行的NAN，如下：

```python
ret = price.pct_change().dropna()
ret.columns = ['ret1', 'ret2', 'ret3']
ret.head()
```

### 接下来我们正式进入seaborn作图环节
    1. sns.distplot()   作出一列数据的频率直方图，同时画出拟合曲线

```python
#比如画出第一只股票，浦发银行的频率直方图
sns.distplot(ret['ret1'])
```

    2.sns.jointplot() 做出两列时间序列的相关图，包括各自的频率直方图，拟合，两个时间序列的散点图，回归直线，置信区间等等

```python
sns.jointplot('ret1','ret2',ret[['ret1', 'ret2']], kind = 'reg') #kind可调节，比如换成 'scatter'就只画散点图
```

```python
sns.jointplot('ret1','ret2',ret[['ret1', 'ret2']], kind = 'kde') #kind可调节，比如换成 'kde'就画等高线图
```

    3. sns.pairplot()  #画出好多只股票之间的相关图，一行代码，简洁实用！

```python
sns.pairplot(ret)
```

    4. sns.boxplot()   #画箱图，比如说画出股票的收盘价的箱图
    5. sns.violinplot  #箱图的另一种表示方法

```python
sns.boxplot(price)
```

```python
sns.violinplot(price)
```
