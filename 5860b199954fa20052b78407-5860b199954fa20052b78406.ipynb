# 基于聚丙烯期货的日内交易策略

**1.前言**
前段时间关注了一个人人宽客的微信公众号，上个星期微信公众号分享了一个关于聚丙烯期货日内交易的一个策略。从文中来看表现不错，本文将其在优矿上进行复现。

**2.策略**
策略的本质是个趋势策略，在每天9点04分判断方向，14点58分平仓，具体逻辑如下：
* 入场：若9点03分的开盘价/今日的开盘价-1 > ranges/1000，开多仓；反之开空仓。
* 出场：每日14点58分平仓。
* 止损：若持有多头头寸，且当前收盘价 < 入场价\*(1000-stoploss\)/1000，平仓出场；若持有空头头寸，且当前收盘价 > 入场价\*(1000+stoploss)/1000，平仓出场。

但是不管怎样，为了避免日内过大的波动，本文将其设为上午11点29分平仓。

**3.可行性分析**
文中说从日内波动性、趋势度、最小变动价位与合约价值比这三个衡量期货品种日内盈利能力的指标来看，聚丙烯在目前商品期货品种中最适合日内交易。以上周一PP1705（12月19日）为例，在低开后迅速走低，仅在9点半至10点间小幅反弹后一路走低至收盘跌停，当日有效波幅超过4%，是典型日内策略盈利的一天。我们可以将当日K线图画出来看下。当然，其实也不尽然。

```python
universe = 'PP1705'                          # 策略期货合约
start = '2016-12-20'                       # 回测开始时间
end  = '2016-12-20'                        # 回测结束时间
capital_base = 1000000                     # 初试可用资金
refresh_rate = 1                          # 调仓周期
freq = 'm'                                 # 调仓频率：m-> 分钟；d-> 日

data = {'PP1705':[]}

def initialize(futures_account):           # 初始化虚拟期货账户，一般用于设置计数器，回测辅助变量等。
    pass

def handle_data(futures_account):          # 回测调仓逻辑，每个调仓周期运行一次，可在此函数内实现信号生产，生成调仓指令。
    symbol = universe
    if futures_account.current_time == '09:00:00':
        data['PP1705'].append(get_symbol_history(symbol=symbol, field='', time_range=226)[symbol])
                    
bt, perf = qf.backtest.backtest(universe=universe, start=start, end=end, initialize=initialize, 
                                handle_data=handle_data, capital_base=capital_base, refresh_rate=refresh_rate,
                                freq=freq)
quotes = data['PP1705'][0]
```

```python
def plot_k(quotes, sec):
    __color_balck__= '#000000'
    __color_green__= '#00FFFF'
    __color_purple__ = '#9900CC'

    fig = plt.figure(figsize=(10,4))
    fig.set_tight_layout(True)
    
    ax1 = fig.add_axes([0, 1, 1, 1])
    ax1.set_title(quotes.ix[0, 'tradeDate'])
    ax1.set_axisbelow(True)
    ax2 = fig.add_axes([0, 0.35, 1, 0.5], axis_bgcolor='w')
    ax2.set_axisbelow(True)
    ax1.grid(True)
    ax2.grid(True)
    ax1.set_xlim(-1, len(quotes)+1)
    ax2.set_xlim(-1, len(quotes)+1)

    for i in range(len(quotes)):
        close_price = quotes.ix[i, 'closePrice']
        open_price = quotes.ix[i, 'openPrice']
        high_price = quotes.ix[i, 'highPrice']
        low_price = quotes.ix[i, 'lowPrice']
        vol = quotes.ix[i, 'volume']
        trade_date = quotes.ix[i, 'barTime']
        if close_price > open_price:
            ax1.add_patch(patches.Rectangle((i-0.2, open_price), 0.4, close_price-open_price, fill=False, color='r'))
            ax1.plot([i, i], [low_price, open_price], 'r')
            ax1.plot([i, i], [close_price, high_price], 'r')
            ax2.add_patch(patches.Rectangle((i-0.2, 0), 0.4, vol, fill=False, color='r'))
        else:
            ax1.add_patch(patches.Rectangle((i-0.2, open_price), 0.4, close_price-open_price, color='g'))
            ax1.plot([i, i], [low_price, high_price], color='g')
            ax2.add_patch(patches.Rectangle((i-0.2, 0), 0.4, vol, color='g'))
    ax1.set_title(sec, fontproperties=font, fontsize=15, loc='left', color='r')
    ax2.set_title(u'成交量', fontproperties=font, fontsize=15, loc='left', color='r')

    ax1.set_xticks(range(0,len(quotes), 15))
    ax2.set_xticks(range(0,len(quotes), 15)) 
    s1 = ax1.set_xticklabels([quotes.ix[index, 'barTime'] for index in ax1.get_xticks()])
    s1 = ax2.set_xticklabels([quotes.ix[index, 'barTime'] for index in ax2.get_xticks()])
         
    ma5 = pd.rolling_mean(np.array(quotes['closePrice'], dtype=float), window=5, min_periods=0)
    ma10 = pd.rolling_mean(np.array(quotes['closePrice'], dtype=float), window=10, min_periods=0)
    ma20 = pd.rolling_mean(np.array(quotes['closePrice'], dtype=float), window=20, min_periods=0)

    ax1.plot(ma5, color='b')
    ax1.plot(ma10, color='y')
    ax1.plot(ma20, color=__color_purple__)
    
    ax1.annotate('MA5-', xy=(len(quotes)-30, ax1.get_yticks()[-1]), color='b', fontsize=10)
    ax1.annotate('MA10-', xy=(len(quotes)-20, ax1.get_yticks()[-1]), color='y', fontsize=10)
    ax1.annotate('MA20-', xy=(len(quotes)-8, ax1.get_yticks()[-1]), color=__color_purple__, fontsize=10)
    vol5 = pd.rolling_mean(np.array(quotes['volume'], dtype=float), window=5, min_periods=0)
    vol10 = pd.rolling_mean(np.array(quotes['volume'], dtype=float), window=10, min_periods=0)
    ax2.plot(vol5, color='b')
    ax2.plot(vol10, color='y')

    return fig
```

**4.风险分析**
我们可以简单的从最大回撤，年化收益以及胜率等指标来对该策略进行分析。

```python
pv = np.array(bt['portfolio_value'])
print '最大回撤：', perf['max_drawdown']
print '年化收益：', (pv[-1]/pv[0]-1)/len(pv)*252
print '夏普比率：', perf['sharpe']
returns = pv[1:]/pv[:-1]
hit = []
win_streak =''
for i in range(len(pv)-1):
    if pv[i] < pv[i+1]:
        hit.append(1)
        win_streak = win_streak + '1'
    if pv[i] > pv[i+1]:
        hit.append(0)
        win_streak = win_streak + '0'
print '胜率:', (float(sum(hit))/len(hit))
max_win_streak = 0
max_lose_streak = 0
for i in win_streak.split('0'):
    if len(i)>max_win_streak:
        max_win_streak = len(i)
for i in win_streak.split('1'):
    if len(i)>max_lose_streak:
        max_lose_streak = len(i)
print '最大连胜次数:', max_win_streak
print '最大连败次数:', max_lose_streak

def max_draw_down(x, start):
    max_down = 0
    peak = start
    a, b, c = 0, 0, 0
    for i in range(len(x)):
        if x[i] > peak:
            peak = x[i]
            a = i
        else:
            down = x[i]/peak - 1
            if down < max_down:
                b = i
                c = a
                max_down = down
    return -max_down, c, b
draw_down = max_draw_down(pv, 50000)

print '最大回撤持续天数:', draw_down[2]-draw_down[1]+1
```

可以看到胜率没有超过文中所说的高于60%，夏普比率达到2.35，如果将平仓时间设为14点58分，策略表现要差点。
