```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
seclist = DataAPI.EquGet(equTypeCD=u"A",secID=u"",ticker=u"",listStatusCD=u"",field=u"",pandas="1")
seclist['listDate'] = pd.to_datetime(seclist['listDate'])
```

```python
idx = DataAPI.MktIdxmGet(beginDate=u"19900101",endDate=u"20170831",indexID=u"000001.ZICN",ticker=u"",field=u"",pandas="1")
idx['endDate'] = pd.to_datetime(idx['endDate'])
idx = idx.set_index('endDate')['preClosePrice'].sort_index()
```

```python
d1 = seclist.set_index('listDate')['secID'].resample('m').count()
d1 = d1['1995':]
d2 = idx['1995':]
```

```python
fig , (a1,a2) = plt.subplots(2,1,sharex=True,sharey=False,figsize=(20,12))
a1.bar(d1.index.to_pydatetime(),d1) # new stock
a2.plot(d2.index.to_pydatetime(),d2) # price change
a1.set_title('new stock')
a2.set_title('price change')
plt.show()
```

```python
data1 = seclist.set_index('listDate')['secID'].resample('m').count()
data1 = data1['1995':]
data2 = idx['1995':].pct_change()
data1.name = 'new stock'
data2.name = 'price change'
```

```python
data1 = (data1 - data1.mean() ) / data1.std()
data2 = (data2 - data2.mean() ) / data2.std()
```

```python
fig , (ax1,ax2) = plt.subplots(2,1,sharex=True,figsize=(20,12))
ax1.bar(data1.index.to_pydatetime(),data1) # new stock
ax2.bar(data2.index.to_pydatetime(),data2) # price change
ax1.set_title('new stock')
ax2.set_title('price change')
plt.show()
```
