```python
import numpy as np
import pandas as pd
import re
import datetime
```

```python
original_df = pd.read_csv('test/stockcode.csv')
original_df.head()
```

```python
# 全部字段去重，去掉tradedate造成的重复
original_df = original_df.drop('tradeDate',axis = 1)
original_df = original_df.drop_duplicates() 
original_df = original_df.reset_index(drop = True)
original_df.size
```

```python
# 提取股票code，并且补全code前面为0部分，并且保存为str类型
i = 0
new_code = []
for temp in pd.isnull(original_df['stockcode']):    
    if temp == False:
        code = original_df['stockcode'][i]
        code_str = str(int(code))
        while code<100000:
            code = code*10
            code_str = '0' + code_str
        new_code.append(code_str)
    else:
        new_code.append('0')
    i = i+1   
original_df['new_stockcode']=new_code
```

```python
original_df.to_csv('stock_sheet.csv')
```

```python
# 日频数据提取过多，提取目标factor
Dfactor_name_df = pd.read_csv('factor_name.csv')
Dfactor_name_list = list(Dfactor_name_df[Dfactor_name_df.columns[0]])
Dfactor_name_list.remove('EP_FY1')
Dfactor_name_list.remove('1/ETOP')
Dfactor_name_list.remove('EP_TTM_Deducted')
Dfactor_name_list = ['secID','ticker','tradeDate']+Dfactor_name_list
print(Dfactor_name_list)
```

```python
# 生成daily数据提取日期
def create_assist_date(datestart = None,dateend = None):
	# 转为日期格式
	datestart=datetime.datetime.strptime(datestart,'%Y%m%d')
	dateend=datetime.datetime.strptime(dateend,'%Y%m%d')
	date_list = []
	date_list.append(datestart.strftime('%Y%m%d'))
	while datestart<dateend:
		# 日期叠加一天
	    datestart+=datetime.timedelta(days=+1)
	    # 日期转字符串存入列表
	    date_list.append(datestart.strftime('%Y%m%d'))
	return date_list
```

```python
# 按年获取每日数据保存并及时删除，否则内存不够用
def Daily_Factor(start,end,code_list,factor_name):
    MktStockFactorsOneDayGet_df = pd.DataFrame()
    print('time: '+start+' - '+end)
    date_list = create_assist_date(datestart = start,dateend = end)
    for date in date_list:
        temp_df = DataAPI.MktStockFactorsOneDayGet(tradeDate=date,secID=u'',ticker=code_list,field=u"",pandas="1")
        if len(temp_df) > 0:
            # print('get:' + str(date))
            temp_df = temp_df[factor_name]
            MktStockFactorsOneDayGet_df = MktStockFactorsOneDayGet_df.append(temp_df)   
    MktStockFactorsOneDayGet_df.to_csv('D\Daily_Factor-'+start+'-'+end+'.csv')
    del MktStockFactorsOneDayGet_df
```

```python
year_list = range(2019,2020)
for year in year_list:
    start = str(year)+'0101'
    end = str(year)+'1231'
    Daily_Factor(start,end,list(original_df['new_stockcode']),Dfactor_name_list)
```

```python
# 可以重启清内存，合并数据每年的数据
original_df = pd.read_csv('stock_sheet.csv')

year_list = range(2011,2020)
Merge_df = pd.DataFrame()
for year in year_list:
    start = str(year)+'0101'
    end = str(year)+'1231'
    Merge_df = Merge_df.append(pd.read_csv('D\Daily_Factor-'+start+'-'+end+'.csv'))
Daily_Factor_Result = pd.merge(original_df,Merge_df,left_on='new_stockcode',right_on='ticker',how='outer')
Daily_Factor_Result.to_csv('D\Factor_Daily.csv')
```
