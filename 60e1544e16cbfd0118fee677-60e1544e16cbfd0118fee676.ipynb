本文通过朴素贝叶斯预测股价：
原理解释下贝叶斯公式，就是用结果推原因的概率，举个例子：有四条生产线生产电视机，第一生产线生产电视机占电视机总数的15%，第二条占20%，第三条30%，第四条35%，其中第一条生产线不合格的概率0.05，第二条不合格率0.04，第三条不合格率0.03，第四条不合格率0.02。已知结果一台电视机是不合格的，求是这台不合格的电视机是第一生产线生产的概率。参考下面解释吧 因为在这里我写不了公式

下面就是演示用朴素贝叶斯公式的流程，在股票预测上其实没有实际意义。应用朴素贝叶斯前提是假设各特征之间相互独立，我选择的特征不满足这个条件。本文目的就是为了学习而已
选择恒瑞医药作为研究标的
选择特征：开盘价 收盘价 最高价 最低价 昨天收盘价 成交量 日换手率 
分类结果 以第二天涨为1。 跌为-1

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
# 选择训练特征
field = ['tradeDate', 'ticker', 'openPrice', 'highestPrice', 'lowestPrice', 'closePrice', 'preClosePrice','turnoverVol', 'turnoverRate','chgPct']
data = DataAPI.MktEqudGet(secID=u"",ticker=u"600276",tradeDate=u"",
                   beginDate=u"20010101",endDate=u"20210101",isOpen="",field=field,pandas="1")
data = pd.DataFrame(data)
# 选择明天的涨跌幅作为分类的依据
data['result'] = data['chgPct'].shift(-1)
data.set_index('tradeDate', inplace=True)
data.index = pd.to_datetime(data.index)
data = data['2001-01-03':'2019-12-29']
data['result'] = np.where(data['result'] > 0, 1, -1)

del data['chgPct']
# data
test = data[['openPrice', 'highestPrice', 'lowestPrice', 'closePrice', 'preClosePrice', 'turnoverVol', 'turnoverRate']]
# test
# 进行数据分割成训练集合和测试集合
x_train, x_test, y_train, y_test = train_test_split(test, data['result'], test_size=0.25)

mlt = MultinomialNB(alpha=1.0)  # alpha 拉普拉斯平滑系数 防止0
mlt.fit(x_train, y_train)
# 预测
mlt.score(x_test, y_test)
```
