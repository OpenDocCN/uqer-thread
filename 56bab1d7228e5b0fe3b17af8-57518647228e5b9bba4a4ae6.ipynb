# 买入三法和卖出三法（K线，均线，KD，量能）

买入三法：  
1.红K突破5日均线  
2.日KD金叉  
3.量能突破5日均量  

卖出三法：  
1.黑K跌破5日均线  
2.日KD死叉  
3.量能跌破5日均量  

此战法融合了K线，均线，KD指标，量能，据说很厉害，按理说也是个OK的思路，但实验下来的效果就呵呵了。。所以不要轻易相信什么神奇的战法，只是去借鉴其中好的部分。。  

日KD的计算比较复杂（其实也还好，看盘软件里面一般是KDJ，但J值用得较少），首先要计算n日的RSV值，即未处理随机指标值，然后再计算K值和D值：  
n日RSV=（Cn－Ln）/（Hn－Ln）* 100  
公式中，Cn为第n日收盘价；Ln为n日内的最低价；Hn为n日内的最高价。  
其次，计算K值与D值：  
当日K值=2/3 * 前一日K值 + 1/3 * 当日RSV  
当日D值=2/3 * 前一日D值 + 1/3 * 当日K值  
若无前一日K值与D值，则可分别用50来代替。
