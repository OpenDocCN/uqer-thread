# 传统技术指标结合机器学习发掘超额收益组合

**思路：量价信息形态信息为特征，使用收益率作为预测目标（未来10，20，40，60日最高收益，收盘收益等，下周期最高收益，下周期收盘收益）**
**举例：**
1. closeIndex_ema5_derivatives：	5日加权ema的方向（导数正负），1表示向上，-1表示向下
2. closeIndex_ema10_derivatives_2：5日加权ema方向转换，比如-1->1为1，表示，5日加权ema方向发生变化，从朝下改为朝上
3. closeIndex_ema10_derivatives_N2:5日加权ema方向转换，从1->-1为1，表示5日加权ema方向发生变化，从朝上改为朝下
4. closeIndex_ema5_10：	5日ema在10日ema上方为1，否则为-1
5. DIF_DEA_relative：macd的dif和dea相对位置，dif在dea上方为1，否则-1
6. DIF_DEA_relative_2:macd的dif和dea相对位置转换，dif向上穿过dea为1，否则0
7. K_D_relative：kdj的k线在d线上方为1，否则-1
等等。

**注意事项**
1，将技术指标信息转化为1和-1等离散取值。
2，预测目标为未来5日内的最高价比率，大于等于5%认为1，否则0（非5日后价格，个人感觉这个较为容易，可与止盈策略搭配使用）
3，考虑到部分交叉点信息，只会体现在交叉当日，需做延迟处理。
如昨日dif金叉dea，则昨日DIF_DEA_relative_2=1，同时今天DIF_DEA_relative_2=1（具体延迟几天视个人感觉）。否则这个信息仅会体现在昨天，利用不够充分。

```python
#参数赋值
start='20140106'
end='20180915'
#样例指数
exampleTicker='399006'
#均线等使用等窗口参数
periodWindow=10

marketIndexDataAll=pd.DataFrame();

#当前所有数据维度
tickerAll=DataAPI.MktIdxdGet(indexID=u"",ticker=u"",tradeDate=u"20140106",beginDate=u"",endDate=u"",exchangeCD=u"XSHE,XSHG",field=['ticker'],pandas="1")
for exampleTickerList in tickerAll.values.tolist()[0:50]:
    exampleTicker=exampleTickerList[0]

    #获取指数行情数据
    marketIndexData=DataAPI.MktIdxdGet(indexID=u"",ticker=exampleTicker,tradeDate='',beginDate=start,endDate=end,field=['tradeDate','closeIndex','turnoverVol','lowestIndex','highestIndex'],pandas="1")
    #列tradeDate修改为日期类型并且设置为索引
    marketIndexData['tradeDate'] = pd.to_datetime(marketIndexData['tradeDate'])
    marketIndexData.set_index("tradeDate", inplace=True)

    ###添加特征，均线相关
    #加权移动平均,导数
    marketIndexData['closeIndex_ema5']=pd.ewma(marketIndexData['closeIndex'], span=5)
    marketIndexData['closeIndex_ema5_derivatives']=marketIndexData['closeIndex_ema5'].diff().apply(lambda x:1 if x>0 else -1).fillna(0)
    marketIndexData['closeIndex_ema5_derivatives_2']=(marketIndexData['closeIndex_ema5_derivatives']+marketIndexData['closeIndex_ema5_derivatives'].diff()/2).apply(lambda x:1 if x==2 else 0).rolling(2).max().fillna(0).apply(lambda x:int(x) if not np.isnan(x) else 0)
    marketIndexData['closeIndex_ema5_derivatives_N2']=(marketIndexData['closeIndex_ema5_derivatives']+marketIndexData['closeIndex_ema5_derivatives'].diff()/2).apply(lambda x:1 if x==-2 else 0).rolling(2).max().fillna(0).apply(lambda x:int(x) if not np.isnan(x) else 0)

    marketIndexData['closeIndex_ema10']=pd.ewma(marketIndexData['closeIndex'], span=10)
    marketIndexData['closeIndex_ema10_derivatives']=marketIndexData['closeIndex_ema10'].diff().apply(lambda x:1 if x>0 else -1).fillna(0)
    marketIndexData['closeIndex_ema10_derivatives_2']=(marketIndexData['closeIndex_ema10_derivatives']+marketIndexData['closeIndex_ema10_derivatives'].diff()/2).apply(lambda x:1 if x==2 else 0).rolling(4).max().fillna(0).apply(lambda x:int(x) if not np.isnan(x) else 0)
    marketIndexData['closeIndex_ema10_derivatives_N2']=(marketIndexData['closeIndex_ema10_derivatives']+marketIndexData['closeIndex_ema10_derivatives'].diff()/2).apply(lambda x:1 if x==-2 else 0).rolling(4).max().fillna(0).apply(lambda x:int(x) if not np.isnan(x) else 0)

    marketIndexData['closeIndex_ema20']=pd.ewma(marketIndexData['closeIndex'], span=20)
    marketIndexData['closeIndex_ema20_derivatives']=marketIndexData['closeIndex_ema20'].diff().apply(lambda x:1 if x>0 else -1).fillna(0)
    marketIndexData['closeIndex_ema20_derivatives_2']=(marketIndexData['closeIndex_ema20_derivatives']+marketIndexData['closeIndex_ema20_derivatives'].diff()/2).apply(lambda x:1 if x==2 else 0).rolling(4).max().fillna(0).apply(lambda x:int(x) if not np.isnan(x) else 0)
    marketIndexData['closeIndex_ema20_derivatives_N2']=(marketIndexData['closeIndex_ema20_derivatives']+marketIndexData['closeIndex_ema20_derivatives'].diff()/2).apply(lambda x:1 if x==-2 else 0).rolling(4).max().fillna(0).apply(lambda x:int(x) if not np.isnan(x) else 0)

    marketIndexData['closeIndex_ema60']=pd.ewma(marketIndexData['closeIndex'], span=60)
    marketIndexData['closeIndex_ema60_derivatives']=marketIndexData['closeIndex_ema60'].diff().apply(lambda x:1 if x>0 else -1).fillna(0)

    #均线相对位置差异,0-5.5-10,5-20,10-20,转为正负（1，-1）信息
    marketIndexData['closeIndex_ema0_5']=(marketIndexData['closeIndex']-marketIndexData['closeIndex_ema5']).apply(lambda x:1 if x>0 else -1)
    marketIndexData['closeIndex_ema5_10']=(marketIndexData['closeIndex_ema5']-marketIndexData['closeIndex_ema10']).apply(lambda x:1 if x>0 else -1)
    marketIndexData['closeIndex_ema5_20']=(marketIndexData['closeIndex_ema5']-marketIndexData['closeIndex_ema20']).apply(lambda x:1 if x>0 else -1)
    marketIndexData['closeIndex_ema10_20']=(marketIndexData['closeIndex_ema10']-marketIndexData['closeIndex_ema20']).apply(lambda x:1 if x>0 else -1)

    # #近（前）n日最高点信息（已实现收益）
    # marketIndexData['closeIndex_max2']=pd.rolling_max(marketIndexData['closeIndex'], window=2)
    # marketIndexData['closeIndex_max5']=pd.rolling_max(marketIndexData['closeIndex'], window=5)
    # 验证相对位置数据正确性
    # print marketIndexData.columns

    ###预测信息
    #未来5，10，20日最高价，收盘价,TODO EMA的最高价
    marketIndexData['closeIndex_max5_future']=((marketIndexData['closeIndex'].rolling(5).max().shift(-5)-marketIndexData['closeIndex'])/marketIndexData['closeIndex']).apply(lambda x:int(x*100) if not np.isnan(x) else 0)
    marketIndexData['closeIndex_max10_future']=((marketIndexData['closeIndex'].rolling(10).max().shift(-10)-marketIndexData['closeIndex'])/marketIndexData['closeIndex']).apply(lambda x:int(x*100) if not np.isnan(x) else 0)
    marketIndexData['closeIndex_max20_future']=((marketIndexData['closeIndex'].rolling(20).max().shift(-20)-marketIndexData['closeIndex'])/marketIndexData['closeIndex']).apply(lambda x:int(x*100) if not np.isnan(x) else 0)

    marketIndexData['closeIndex_close5_future']=((marketIndexData['closeIndex'].shift(-5)-marketIndexData['closeIndex'])/marketIndexData['closeIndex']).apply(lambda x:int(x*100) if not np.isnan(x) else 0)
    marketIndexData['closeIndex_close10_future']=((marketIndexData['closeIndex'].shift(-10)-marketIndexData['closeIndex'])/marketIndexData['closeIndex']).apply(lambda x:int(x*100) if not np.isnan(x) else 0)
    marketIndexData['closeIndex_close20_future']=((marketIndexData['closeIndex'].shift(-20)-marketIndexData['closeIndex'])/marketIndexData['closeIndex']).apply(lambda x:int(x*100) if not np.isnan(x) else 0)
    # 验证相对位置数据正确性
    # print marketIndexData.columns
    # print marketIndexData.head(10)


    #添加macd线信息
    #2种信息。
    # 1，dif导数信息
    # 2，dea导数信息
    # 3，dif和dea的相对位置。
    # 4, dif和dea的交叉信息，上穿下-2，下穿上2。
    DIF,DEA,MACD = talib.MACD(marketIndexData['closeIndex'].values, fastperiod=12, slowperiod=26, signalperiod=20)
    marketIndexData['DIF_derivatives']=DIF
    marketIndexData['DIF_derivatives']=marketIndexData['DIF_derivatives'].diff().apply(lambda x:1 if x>0 else -1)
    marketIndexData['DEA_derivatives']=DEA
    marketIndexData['DEA_derivatives']=marketIndexData['DEA_derivatives'].diff().apply(lambda x:1 if x>0 else -1)

    marketIndexData['DIF_DEA_relative']=DIF-DEA
    marketIndexData['DIF_DEA_relative']=marketIndexData['DIF_DEA_relative'].apply(lambda x:1 if x>0 else -1)
    marketIndexData['DIF_DEA_relative_2']=(marketIndexData['DIF_DEA_relative']+marketIndexData['DIF_DEA_relative'].diff()/2).apply(lambda x:1 if x==2 else 0).rolling(5).max().apply(lambda x:int(x) if not np.isnan(x) else 0)
    marketIndexData['DIF_DEA_relative_N2']=(marketIndexData['DIF_DEA_relative']+marketIndexData['DIF_DEA_relative'].diff()/2).apply(lambda x:1 if x==-2 else 0).rolling(5).max().fillna(0).apply(lambda x:int(x) if not np.isnan(x) else 0)

    marketIndexData_2=marketIndexData[['DIF_DEA_relative','DIF_derivatives','DEA_derivatives','DIF_DEA_relative_2','DIF_DEA_relative_N2']]
    # print marketIndexData[['DIF_DEA_relative','DIF_derivatives','DEA_derivatives','DIF_DEA_relative_2','DIF_DEA_relative_N2']].head(10)

    #添加kdj信息
    # 1，k线和d线相对位置
    # 2，d线大于80
    # 3，d线小于20
    k,d=talib.STOCH(np.array(marketIndexData['highestIndex']),np.array(marketIndexData['lowestIndex']),np.array(marketIndexData['closeIndex']),
      fastk_period=9,slowk_period=3,slowk_matype=0,slowd_period=3,slowd_matype=0)
    marketIndexData['K_D_relative']=k-d
    marketIndexData['K_D_relative']=marketIndexData['K_D_relative'].diff().apply(lambda x:1 if x>0 else -1).fillna(0)
    marketIndexData['K_80_relative']=k
    marketIndexData['K_80_relative']=marketIndexData['K_80_relative'].apply(lambda x:1 if x>=80 else 0).fillna(0)
    marketIndexData['K_20_relative']=k
    marketIndexData['K_20_relative']=marketIndexData['K_20_relative'].apply(lambda x:1 if x<=20 else 0).fillna(0)
    # print marketIndexData[['K_D_relative','K_80_relative','K_20_relative']].head(10)

    #添加RSA信息
    #1，大于80
    #2，小于20
    rsi=talib.RSI(np.array(marketIndexData['closeIndex']), timeperiod=12)  
    marketIndexData['RSA_80_relative']=rsi
    marketIndexData['RSA_80_relative']=marketIndexData['RSA_80_relative'].apply(lambda x:1 if x>=80 else 0).fillna(0)
    marketIndexData['RSA_20_relative']=rsi
    marketIndexData['RSA_20_relative']=marketIndexData['RSA_20_relative'].apply(lambda x:1 if x<=20 else 0).fillna(0)
    
  
    marketIndexData_2=marketIndexData[[ 'closeIndex_max5_future','closeIndex_ema5_derivatives',
                                   'closeIndex_ema5_derivatives_2',
                                    'closeIndex_ema5_derivatives_N2',
                                'closeIndex_ema10_derivatives',
                                'closeIndex_ema10_derivatives_2',
                                'closeIndex_ema10_derivatives_N2',
                                'closeIndex_ema20_derivatives',
                                'closeIndex_ema20_derivatives_2',
                                'closeIndex_ema20_derivatives_N2',
                                   'closeIndex_ema60_derivatives', 
                                   u'closeIndex_ema5_10', 
                                   u'closeIndex_ema5_20',
                                   u'closeIndex_ema10_20',
                                   'DIF_derivatives',
                                    'DEA_derivatives',
                                    'DIF_DEA_relative',
                                    'DIF_DEA_relative_2',
                                    'DIF_DEA_relative_N2',
                                    'K_D_relative',
                                    'K_80_relative',
                                    'K_20_relative',
                                    'RSA_80_relative',
                                    'RSA_20_relative'
                                 ]]
    marketIndexDataAll = pd.concat([marketIndexDataAll, marketIndexData_2], axis=0)

```
