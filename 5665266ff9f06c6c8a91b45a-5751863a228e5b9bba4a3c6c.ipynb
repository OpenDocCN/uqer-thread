## 1 日期和时间数据类型及工具

datetime是使用的最多的数据类型，以毫秒形式存储日期和时间：

```python
from datetime import datetime
now = datetime.now()
now
```

```python
now.year, now.month, now.day
```

```python
delta = datetime(2011, 1, 7) - datetime(2008, 6, 24, 8, 15)
delta
```

```python
delta.days
```

```python
from datetime import timedelta
start = datetime(2011, 1, 7)
start + timedelta(12)
```

```python
start - 2 * timedelta(12)
```

datetime可以转换为文本：

```python
stamp = datetime(2011, 1, 3)
str(stamp)
```

'%Y-%m-%d'是时间格式，分别表示为年月日：

```python
stamp.strftime('%Y-%m-%d')
```

字符串也可以转换为datetime：

```python
value = '2011-01-03'
datetime.strptime(value, '%Y-%m-%d')
```

`datetime.strptime()`是通过已知格式进行日期解析的最佳方式，但是每次都要编写格式定义是很麻烦的。dateutill这个第三方包中的parser.parse方法集合可以解析所以人类能够理解的日期表示形式：

```python
parse('Jan 31, 1997 10:45 PM')
```

国际通用格式中，日通常在出现在月的前面，传入dayfirst=True即可解决：

```python
parse('6/12/2011', dayfirst=True)
```

### 3.1 生成日期范围

`pandas.date_range`用于生成制定长度的DatetimeIndex：

```python
index = pd.date_range('4/1/2012', '6/1/2012')
index
```

该方法默认产生日频索引，如果只传入起始日期或结束日期，则还需要传入时期数：

通过对freq设定不同的参数（详见3.2），可以得到不同的时间索引，如'BM'就是产生由每月最后一个工作日组成的日期索引：

### 3.3 移动（超前和滞后）数据

```python
ts = Series(np.random.randn(4), index=pd.date_range('1/1/2000', periods=4, freq='M'))
ts
```

使用`.shift()`方法可以实现单纯的前移或后移操作：

```python
ts.shift(2)
```

```python
ts.shift(-2)
```

可以使用这种方法计算时间序列的百分比变化：

```python
ts / ts.shift(1) - 1
```

单纯的移位操作不会修改索引，所以部分数据会被丢弃。因此，如果数据频率已知，可以通过freq对时间索引进行位移而不是对数据进行位移：

```python
ts.shift(2, freq='M')
```

```python
ts.shift(3, freq='D')
```

```python
ts.shift(1, freq='3D')
```

```python
ts.shift(1, freq='90T')
```

### 4.2 按季度计算的时期频率

不同国家对于财年的规定可能有所不同，因此季度数据有时会显得比较麻烦。

如果某一财年是以1月份结束的，则其财年四季度就是11月到1月：

```python
p = pd.Period('2012Q4', freq='Q-JAN')
p
```

```python
p.asfreq('D', 'start')
```

```python
p.asfreq('D', 'end')
```

例子：获取该季度倒数第二个工作日下午4时的时间戳：

```python
p4pm = (p.asfreq('B', 'e') - 1).asfreq('T', 's') + 16 * 60
p4pm
```

```python
p4pm.to_timestamp()
```

```python
rng = pd.period_range('2011Q3', '2012Q4', freq='Q-JAN')
ts = Series(np.arange(len(rng)), index=rng)
ts
```

```python
new_rng = (rng.asfreq('B', 'e') - 1).asfreq('T', 's') + 16 * 60
ts.index = new_rng.to_timestamp()
ts
```

## 5 重采样及频率转换

重采样指的是将时间序列从一个频率转换到另一个频率的处理过程。  
将高频数据聚合到低频率成为降采样。  
将低频率数据转换到高频率成为升采样。

需要使用`.resample()`方法：

```python
rng = pd.date_range('1/1/2000', periods=100, freq='D')
ts = Series(randn(len(rng)), index=rng)
ts.resample('M', how='mean')
```

```python
ts.resample('M', how='mean', kind='period')
```

>`.resample()`方法的重要参数：

|参数|说明|
|:----------|:----------|
|freq|表示重采样频率的字符串，如'M'、'5min'、'second(15)'|
|how|用于产生聚合值的函数名或数组函数，如'mean'、'first'、'last'、'median'、'ohlc'、'max'、'min'|
|fill\_method　　|升采样时如何插值，如'ffill'、'bfill'，默认不插值|
|closed|在降采样中，各时间段的哪一端是闭合的|
|label|在降采样中，如何设置聚合值的标签|

### 5.1 降采样

例如将1分钟数据转换为5分钟数据：

```python
rng = pd.date_range('1/1/2000', periods=12, freq='T')
ts = Series(np.arange(12), index=rng)
ts
```

```python
ts.resample('5min', how='sum')
# note: output changed (as the default changed from closed='right', label='right' to closed='left', label='left'
```

```python
ts.resample('5min', how='sum', closed='left')
```

```python
ts.resample('5min', how='sum', closed='left', label='left')
```

OHLC重采样，金融领域中常用，计算一段区间内序列的四个值：开盘、最高、最低、收盘。

```python
ts.resample('5min', how='ohlc')
```

### 5.2 升采样和插值

一个周度的DataFrame：

```python
frame = DataFrame(np.random.randn(2, 4),
                  index=pd.date_range('1/1/2000', periods=2, freq='W-WED'),
                  columns=['Colorado', 'Texas', 'New York', 'Ohio'])
frame
```

## 7 移动窗口函数

```python
close_px = close_px.asfreq('B').fillna(method='ffill')
```

```python
close_px.AAPL.plot()
pd.rolling_mean(close_px.AAPL, 250).plot()
```

```python
appl_std250 = pd.rolling_std(close_px.AAPL, 250, min_periods=10)
appl_std250[5:12]
```

```python
appl_std250.plot()
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1, sharex=True, sharey=True, figsize=(12, 7))

aapl_px = close_px.AAPL['2005':'2009']

ma60 = pd.rolling_mean(aapl_px, 60, min_periods=50)
ewma60 = pd.ewma(aapl_px, span=60)

aapl_px.plot(style='k-', ax=axes[0])
ma60.plot(style='k--', ax=axes[0])
aapl_px.plot(style='k-', ax=axes[1])
ewma60.plot(style='k--', ax=axes[1])
axes[0].set_title('Simple MA')
axes[1].set_title('Exponentially-weighted MA')
```

>移动窗口和指数加权函数

|函数|说明|
|:---------|:------------|
|rolling\_count|返回各窗口非NA观测值的数量|
|rolling\_sum|移动窗口的和|
|rolling\_mean|移动窗口的平均值|
|rolling\_median|移动窗口的中位数|
|rolling\_var、rolling\_std|移动窗口的方差和标准差|
|rolling\_skew、rolling\_kurt　　|移动窗口的偏度和峰度|
|rolling\_min、rolling\_max|移动窗口的最小值和最大值|
|rolling\_quantile|移动窗口的指定百分位数/样本分位数位置的值|
|rolling\_corr、rolling\_cov|移动窗口的相关系数和协方差|
|rolling\_apply|对移动窗口应用普通数组函数|
|ewma|指数加权移动平均|
|ewmvar、ewmstd|指数加权移动方差和标准差|
|ewmcorr、ewmcov|指数加权移动相关系数和协方差|

二元移动窗口函数：

```python
spx_px = close_px_all['SPX']
```

```python
spx_rets = spx_px / spx_px.shift(1) - 1
returns = close_px.pct_change()
corr = pd.rolling_corr(returns.AAPL, spx_rets, 125, min_periods=100)
corr.plot()
```

传入一个DataFrame和Series会计算每一列数据与Series之间的函数值：

```python
corr = pd.rolling_corr(returns, spx_rets, 125, min_periods=100)
corr.plot()
```

用户定义的移动窗口函数：

```python
from scipy.stats import percentileofscore
score_at_2percent = lambda x: percentileofscore(x, 0.02)
result = pd.rolling_apply(returns.AAPL, 250, score_at_2percent)
result.plot()
```
