本文选的是000001.XSHE平安银行这只股票（就只是单纯随便选而已），选取了波动率指标：EMV，短期和长期均线指标：EMA5和EMA60，MA5和MA60，动量和反转指标：REVS5和REVS20，成交量指标：VEMA5，财务指标：PE，能量指标：Elder，这12个指标作为输入值，下一日股票的涨跌幅作为输出值。<br>
考虑到样本的容量需要保证，但是太久远的数据可能参考性不大，本文选取2012-2015这三年的输入和输出总共627组样本作为训练集。

```python
DATA=DataAPI.MktStockFactorsDateRangeGet(secID=u"",ticker=u"000001",beginDate=u"20120101",endDate=u"20150101",field=u"tradeDate,PE,Elder,DEA,VEMA5,EMA5,EMV6,EMA60,REVS20,REVS5,MA60,MA5",pandas="1")
P=DataAPI.MktEqudGet(tradeDate=u"",secID=u"",ticker=u"000001",beginDate=u"20120101",endDate=u"20150101",isOpen="",field=u"",pandas="1").closePrice
R=P.shift(-1)-P
#R=P.shift(-1)
#R=Series([1 if i>0 else 0 for i in list(R)])
DATA=pd.concat([DATA,R],axis=1,ignore_index=True)
DATA=DATA.dropna(axis=0)
DATA
```

这里对数据进行了中心化处理，保证不同参数的可比性。

```python
for fac in DATA.columns[1:]:
    DATA[fac]=(DATA[fac]-np.mean(DATA[fac]))/np.std(DATA[fac])
DATA
```

```python
U=zip(list(DATA[1]),list(DATA[2]),list(DATA[3]),list(DATA[4]),list(DATA[5]),list(DATA[6]),list(DATA[7]),list(DATA[8]),list(DATA[9]),list(DATA[10]),list(DATA[11]),list(DATA[12]))
M=[list(i) for i in U]
M
```

这里定义了Sigmoid函数其作用是将取值为R的输入映射到[-1,1]上，以满足神经网络模型的输入需要。

```python
def Sigmoid(x):
    u=1/(2.71**(-x)+1)
    return u
```

```python
Data_set=[]
for i in M:
    i=[Sigmoid(j) for j in i]
    Data_set.append(i)
Data_set
```

```python
from pybrain.tools.shortcuts import buildNetwork
from pybrain.datasets import SupervisedDataSet
Dataset=SupervisedDataSet(11,1)
for i in range(len(Data_set)):
    Dataset.appendLinked(Data_set[i][:-1],Data_set[i][-1])
print Dataset
```

这一步是训练模型，本文设定的模型除了输入和输出层外，含有两个隐藏层，每个隐藏层含有13个单元，并且设定了迭代次数为500次，理论上迭代次数应该越多越好，但实际上后面的迭代带来的误差缩小已经微乎其微。<br>
另外这里的momentum的大小是解跳出局部最优的步长，太大就很难收敛，太小就容易陷入局部最优。weightdecay指的是单次权重改变的步长，太大也会不容易收敛，太小就比较慢。<br>
此模型训练结果每次都不一样，可以多训练几次找到误差相对较小的保存下来。

```python
from pybrain.datasets import ClassificationDataSet
from pybrain.utilities import percentError
from pybrain.tools.shortcuts import buildNetwork
from pybrain.supervised.trainers import BackpropTrainer
from pybrain.structure.modules import SoftmaxLayer
fnn = buildNetwork( Dataset.indim, 13,13, Dataset.outdim)
trainer=BackpropTrainer(fnn,Dataset,momentum=0.01,verbose=True,weightdecay=0.0001)
trainer.trainEpochs(epochs=500)
```

将模型运用于测试集的效果。

```python
U=fnn.activateOnDataset(Dataset)
fig = plt.figure(figsize=(14,10))
fig.set_tight_layout(True)
ax1 = fig.add_subplot(211)
ax1.plot(U)
ax2 = ax1.twinx()
h=[Sigmoid(j) for j in DATA[12]]
ax2.plot(h,'r')
plt.show()
```

本文选取2015-2016年作为测试集，将模型用于测试集的效果，可以看出还是有一定的跟踪效果的。

```python
DATA2=DataAPI.MktStockFactorsDateRangeGet(secID=u"",ticker=u"000001",beginDate=u"20150101",endDate=u"20160101",field=u"tradeDate,PE,Elder,DEA,VEMA5,EMA5,EMV6,EMA60,REVS20,REVS5,MA60,MA5",pandas="1")
P2=DataAPI.MktEqudGet(tradeDate=u"",secID=u"",ticker=u"000001",beginDate=u"20150101",endDate=u"20160101",isOpen="",field=u"",pandas="1").closePrice
R2=P2.shift(-1)-P2
#R=Series([1 if i>0 else 0 for i in list(R)])
DATA2=pd.concat([DATA2,R2],axis=1,ignore_index=True)
DATA2=DATA2.dropna(axis=0)
for fac in DATA2.columns[1:]:
    DATA2[fac]=(DATA2[fac]-np.mean(DATA2[fac]))/np.std(DATA2[fac])
U2=zip(list(DATA2[1]),list(DATA2[2]),list(DATA2[3]),list(DATA2[4]),list(DATA2[5]),list(DATA2[6]),list(DATA2[7]),list(DATA2[8]),list(DATA2[9]),list(DATA2[10]),list(DATA2[11]),list(DATA2[12]))
M2=[list(i) for i in U2]
Data_set2=[]
for i in M2:
    i=[Sigmoid(j) for j in i]
    Data_set2.append(i)
I=[]
O=[]
for D in Data_set2:
    I.append(D[:-1])
    O.append(D[-1])
FO=[]
for i in I:
    FO.append(fnn.activate(i))
fig = plt.figure(figsize=(14,10))
ax1 = fig.add_subplot(211)
ax2=ax1.twinx()
ax1.plot(O)
ax2.plot(FO,'r')
plt.grid()
plt.show()
```
