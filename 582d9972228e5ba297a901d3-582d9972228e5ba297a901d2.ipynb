## 1. A股的市值异象
所谓市值异象，是指小市值股票相对大市值股票具有显著溢价。然而在靠近年底的12月份，市值因子的选股效果大打折扣。小市值因子会在年底的时候失效，这与市场风险偏好变化，投资者行为等因素息息相关。投资者基于流动性配置需求的投资行为，是引起该效应的潜在原因之一。 本文主要参考海通证券的《市值异象的岁末效应》研报。

下面通过简单的统计来证明这种情况的存在：
以2006年1月4号至2016年11月11号的时间窗口为例，按照**前一日总市值由低到高**，将A股分为10组，分别记作D1至D10（D即Decile，代表十分之一）,这里面要剔除两种股票：
1. 停牌的股票
2. 上市不足3个月的新股

下面是冗长的代码部分：

```python
import seaborn
import pandas as pd
import numpy as np
from matplotlib import pylab
import matplotlib.pyplot as plt
from matplotlib.pyplot import *
from CAL.PyCAL import *    # CAL.PyCAL中包含font
from datetime import timedelta,datetime
import datetime
import cPickle
```

### 1.1 下面这个模块主要提供常用的一些小函数，部分来源于knowledge和社区分享

```python
cal = Calendar('China.SSE')

def new_remove(ticker,tradeDate= None,day = 1):
    """
    给定股票列表,去除某个日期前多少个交易日,之后上市的新股
    Args:
        ticker (list of str): 需要进行筛选的股票列表（无后缀）
        tradeDate (datetime): 进行筛选的日期,默认为调用当天
        day (int): 向前漂移的交易日的个数
    Returns:
        list: 去掉新股股票之后的股票列表（无后缀）

    Examples:
        >> universe = set_universe('A')
        >> universe = [stk[0:6] for stk in universe]
        >> universe_without_new = new_remove(universe)
    """
    tradeDate = tradeDate if tradeDate is not None else datetime.datetime.now()
    period = '-' + str(day) + 'B'
    pastDate = cal.advanceDate(tradeDate,period)
    pastDate = pastDate.strftime("%Y-%m-%d")
    ipo_date = DataAPI.SecIDGet(partyID=u"",assetClass=u"e",ticker=ticker,cnSpell=u"",field=u"ticker,listDate",pandas="1")
    remove_list = ipo_date[ipo_date['listDate'] > pastDate]['ticker'].tolist()
    return [stk for stk in ticker if stk not in remove_list]

def get_Atickers(date):
    '''
    功能：获取某一天的全部上市A股ticker（包括上市和暂停上市）
    输入：
        date，yyyymmdd日期
    输出：
        list of tickers
    依赖：
        DataAPI.EquGet
    '''
    
    try:
        universe = DataAPI.EquGet(equTypeCD=u"A",listStatusCD="L,S,DE",field=u"ticker,listDate,delistDate",pandas="1")
    except:
        raise ValueError('DataAPI.EquGet取数据异常！！！')
    universe['listdate'] = universe['listDate'].apply(lambda x: x.replace('-',''))  # 只取date之前上市的股票
    universe['delistdate'] = universe['delistDate'].apply(lambda x: x.replace('-','') if isinstance(x, str) else '99999999')
    universe = universe[(universe['listdate'] <= date) & (universe['delistdate'] > date)]
    tickers = universe['ticker'].tolist()
    return tickers
  
def get_non_halt_tickers(date):
    '''
    功能：获取A股某一天的非停牌股票
    输入：
        date，yyyymmdd型日期
    输出：
        list，这一天非停牌的股票tickers
    依赖：
        需要用到DataAPI：SecHaltGet
    '''
    universe = get_Atickers(date)
    halt_tickers = DataAPI.SecHaltGet(beginDate=date,endDate=date,ticker=universe,field=u"",pandas="1")['ticker'].tolist()
    return list(set(universe) - set(halt_tickers))

        
def get_dates(start_date, end_date, frequency='daily'):
    '''
    功能：输入起始日期和频率，即可获得日期列表（daily包括起始日，其余的都是位于起始日中间的）
    输入参数：
       start_date，开始日期，'xxxxxxxx'形式
       end_date，截止日期，'xxxxxxxx'形式
       frequency，频率，daily为所有交易日，daily1为所有自然日，weekly为每周最后一个交易日，weekly2为每隔两周，monthly为每月最后一个交易日，quarterly为每季最后一个交易日
    输出参数：
       获得list型日期列表，以'xxxxxxxx'形式存储
    PS:
        要用到DataAPI.TradeCalGet！！！
    '''
    
    data = DataAPI.TradeCalGet(exchangeCD=u"XSHG",beginDate=start_date,endDate=end_date,field=u"calendarDate,isOpen,isWeekEnd,isMonthEnd,isQuarterEnd",pandas="1")
    if frequency == 'daily':
        data = data[data['isOpen'] == 1]
    elif frequency == 'daily1':
        pass
    elif frequency == 'weekly':
        data = data[data['isWeekEnd'] == 1]
    elif frequency == 'weekly2':
        data = data[data['isWeekEnd'] == 1]
        data = data[0:data.shape[0]:2]
    elif frequency == 'monthly':
        data = data[data['isMonthEnd'] == 1]
    elif frequency == 'quarterly':
        data = data[data['isQuarterEnd'] == 1]
    else:
        raise ValueError('调仓频率必须为daily/daily1/weekly/weekly2/monthly/quarterly！！！')
    date_list = map(lambda x: x[0:4]+x[5:7]+x[8:10], data['calendarDate'].values.tolist())
    return date_list


def get_monthend(date_str='20160104'):
    """获取该日的上个月的最后一个交易日
    比如输入20161115，那么输出20161031"""
    begin_date = datetime.strptime(date_str, '%Y%m%d') - timedelta(days=1)  # 排除今天
    end_date = begin_date - 100 * timedelta(days=1)
    
    begin_day_str = begin_date.strftime('%Y-%m-%d')
    end_day_str = end_date.strftime('%Y-%m-%d')
    
    return get_dates(begin_day_str, end_day_str,'monthly')[-1]
```

### 1.2 这个模块主要把研究中需要用到的数据落地为pickle和csv文件，为了后续的研究比较方便快捷

1. 如果要跑通下面的代码，需要在优矿Data中创建qualified_tickers文件夹
2. 主要运行dump_qualified_tickers和get_return_every_decile两个函数即可
2. dump_qualified_tickers这个函数需要先跑一下，落地每天的股票池
3. quantile_return.csv保存的就是分组每日的收益，后面主要用到这个数据

```python
def get_qualified_tickers(day):
    """获取当天合格的A股股票list，
       主要剔除停牌和上市不足3个月的
    """ 
    non_halt_list = get_non_halt_tickers(day) # 非停牌股票
    return new_remove(non_halt_list, day, 60)  # 滤除上市不超过3个月的新股

def dump_qualified_tickers(begin='20060104', end='20161111'):
    '''落地每天的股票池，因为直接读API速度会比较慢，
       落地后读取速度会加快很多，方便研究
       每天的list作为一个pickle文件，按天存放，都存放在qualified_tickers目录下
    '''
    trading_day_list = get_dates(begin, end)
    for trading_day in trading_day_list:
        file_path = 'qualified_tickers/'+trading_day + '.txt'  
        tickers = get_qualified_tickers(trading_day)
        write(file_path, cPickle.dumps(tickers), encoding = 'utf-8')
        
def get_qualified_tickers(day):
    """读取缓存的每天合格的股票池"""
    file_path = 'qualified_tickers/' + day + '.txt'
    return cPickle.loads(str(read(file_path)))

# dump_qualified_tickers()   # 用来dump数据，估计需要3个小时左右

def get_return_every_decile(begin='20060104', end='20161111'): 
    """按照市值分成10份，每份的收益都以字典的形式存储下来。字典的key是日期，值是每个分位的改天的平均收益
       嗯，截止日默认为光棍节
    """
    ret_dict = {} 
    trading_day_list = get_dates(begin, end)
    for index,trading_day in enumerate(trading_day_list):
        # 获取昨日的市值
        previous_trading_day = trading_day_list[index-1] if index > 0 else trading_day_list[0]
        mkt = DataAPI.MktEqudGet(tradeDate=previous_trading_day,secID=u"",ticker='',isOpen="",field=u"ticker,marketValue",pandas="1")
        
        # 获取当日收益(考虑现金红利再投资)
        daily_return = DataAPI.EquRetudGet(listStatusCD=u"L",secID=u"",ticker='',beginDate=trading_day,endDate=trading_day,isChgPctl=u"",field=u"ticker,dailyReturnReinv",pandas="1")
        
        mkt = mkt.merge(daily_return,how='inner',on='ticker')
        
        universe = get_qualified_tickers(trading_day)
        mkt = mkt[mkt['ticker'].isin(universe)]
        mkt['lable'] = pd.qcut(mkt['marketValue'], 10, labels=False) # 按照市值划分为10个等分，0代表市值最小的部分，9代表市值最大的部分

        ret_dict[trading_day] = mkt.groupby("lable")['dailyReturnReinv'].mean()
    return ret_dict
    
ret_dict = get_return_every_decile()  # 将字典转成dataframe,并且落地，供后续研究，运行时间需要20分钟
df = pd.DataFrame.from_dict(ret_dict,orient='index')
df.columns = ['D'+str(item+1) for item in df.columns]
df.to_csv('quantile_return.csv')
```

```python
qunatile_df = pd.read_csv('quantile_return.csv',index_col=0) # 第一列的时间作为index
quantile_return_se = qunatile_df.mean()  # 计算每日的平均收益
sharp_se = qunatile_df.mean()  / qunatile_df.std() * np.sqrt(250) # 计算夏普值


# --------------下面是画图部分------------------------
fig = plt.figure(figsize=(12, 6))
ax1 = fig.add_subplot(111)
ax2 = ax1.twinx()

ax1.bar(range(0,len(quantile_return_se.index)), quantile_return_se.values, align='center', color='r', width=0.5, label=u'日均收益率，左轴')
ax2.plot(range(0,len(sharp_se.index)), sharp_se.values,'o-g', label=u'夏普比率，右轴')
ax1.set_xticks(range(0,11))
ax1.set_xticklabels(quantile_return_se.index, fontproperties=font, fontsize=14)
ax2.set_xticks(range(0,11))
ax2.set_ylim(0.0,2.2)
ax2.set_xticklabels(quantile_return_se.index, fontproperties=font, fontsize=14)
ax1.set_title(u"2006～2016日均收益与夏普比率（按市值从小到大）", fontproperties=font, fontsize=16)
ax1.legend(loc=2,prop=font)
ax2.legend(loc=1,prop=font)
ax2.grid()
```

### 1.3 小市值效应的原因：
1. 从资产定价角度看，小市值股票具有高波动，低流动性等特征，需要更高的收益补偿
2. 从市场机制角度看，A 股市场严格的IPO审批制度，使得小市值公司具有壳价值
3. 从经济转型角度看，部分小市值公司具有题材新颖，高成长空间的特点
4. 个人觉得这个跟A股散户居多，风险偏好较大也有关系
     
基于以上的原因，小市值的股票具有高风险高回报的特性

## 2. 市值异象的岁末效应
上一节中可以看到小市值股票对于大市值股票的很明显的超额收益，然而这个结论并不总是成立的，在每年的12月，会有一个反转。

### 2.1 首先可以统计一下每个月的小市值相对大市值的胜率：

```python
success_rate_df = qunatile_df.copy()  
success_rate_df.index = success_rate_df.index.astype(str).to_datetime()
success_rate_df = success_rate_df[success_rate_df.index<'20160101']   # 因为2016年的数据不全，剔除2016年的数据

success_rate_df['success_count'] = (success_rate_df['D1']-success_rate_df['D10']).apply(lambda x: 1 if x>0 else 0)  # 小市值对大市值的胜率
success_rate_se = success_rate_df.groupby(lambda x: x.month)['success_count'].sum() / success_rate_df.groupby(lambda x: x.month).size()

# ----------------下面是画图部分-----------------------
fig = plt.figure(figsize=(12, 6))
ax1 = fig.add_subplot(111)
ax1.plot(range(0,len(success_rate_se.index)), success_rate_se.values,'o-b', label=u'日胜率')
ax1.legend(loc=1,prop=font)
ax1.set_ylim(0.4,0.8)
ax1.set_xlim(-1,12)
ax1.set_xticks(range(0,len(success_rate_se.index)))
ax1.set_xticklabels(success_rate_se.index, fontproperties=font, fontsize=14)
ax1.set_title(u"小市值对大市值的日胜率（月度）", fontproperties=font, fontsize=16)
```

上面图可以很明显的看到，在12月份，小市值对大市值的胜率并没有小于50%....与研报中的略有差异。但是的确是所有月份中的最低值。


### 2.2 下面来统计小市值相对大市值组合日均超额收益：

```python
exceed_ret_df = qunatile_df.copy()  
exceed_ret_df.index = exceed_ret_df.index.astype(str).to_datetime()

exceed_ret_df['exceed_return'] = exceed_ret_df['D1']-exceed_ret_df['D10']  # 小市值对大市值的超额收益 
exceed_ret_df = exceed_ret_df[exceed_ret_df.index<'20160101']   # 因为2016年的数据不全，剔除2016年的数据

exceed_ret_df
exceed_return_se = exceed_ret_df.groupby(lambda x: x.month)['exceed_return'] .mean()


#----------------下面是画图部分-----------------------
fig = plt.figure(figsize=(12, 6))
ax1 = fig.add_subplot(111)
ax1.bar(range(0,len(exceed_return_se.index)), exceed_return_se.values, align='center', color='r', width=0.5, label=u'日均收益率，左轴')
ax1.set_xticks(range(0,len(exceed_return_se.index)))
ax1.set_xticklabels(exceed_return_se.index, fontproperties=font, fontsize=14)
ax1.set_title(u"小市值对大市值的超额收益（月度）", fontproperties=font, fontsize=16)
```

### 2.3 虚拟变量回归方程
为了更好的判断市值异象岁末效应的显著性，研报中参考了Donald B.Keim的《Size-Related Anomalies and Stock Return Seasonality》论中构建虚拟虚拟变量回归方程。

$R_{t}=\alpha_{12} + \alpha_{1}D_{1t} + \alpha_{2}D_{2t} + \cdots  +\alpha_{11}D_{11t} + \epsilon_{t}$

其中$R_{t}$是小市值相对大市值组合的日收益率时间序列；$D_{it}$为代表各月份的虚拟变量。例如，当$D_{1t} =1$时，其余$D_{it} = 0$时，表示当前交易日为1月，以此类推。若各虚拟变量均为零，则代表当前交易日为12月。


对于上述回归方程，主要关心两个问题：
1. 回归系数是否联合显著，表明大小市值组合的超额收益是否与自然月相关；
2. 截距项$\alpha_{12}$是否显著小于零，表明12月份大市值组合相对小市值组合是否具有显著的超额收益。


回归的时间范围从20060104—～20161111


```python
import statsmodels.api as sm
exceed_ret_df['month'] = exceed_ret_df.index.month
dummy_df = pd.crosstab(exceed_ret_df['exceed_return'], columns=exceed_ret_df['month'])  # 生成月份的虚拟变量的矩阵
dummy_df.columns = ['M'+str(item) for item in dummy_df.columns]  # M 代表月份

x_df = dummy_df[['M1', 'M2', 'M3', 'M4', 'M5', 'M6', 'M7', 'M8', 'M9', 'M10','M11']]
X = sm.add_constant(x_df.values)
Y = dummy_df.index.values
result = sm.OLS(Y,X).fit()

print result.summary()
```

## 3. 总结
1. 从上面的回归结果可以看到，回归分析的结果并不严格支持以上的结论，不知道是不是本人的回归方法有问题，也就是说12月份大市值战胜小市值并非严格显著。

2. 虽然显著性不够强，但是在上面的分析中可以看到，年底大市值的股票的确更值得关注。研报中认为出现这种情况的原因主要是流动性需求。临近年底，投资者对于流动性配置的需求上升，而大市值的股票可以提供充足的流动性。流动性的确会带来溢价，研报中最后使用了非流动性指标，也就是illiquidity这个因子来筛选出流动性比较好的股票，这个因子在社区上已经有大神详细分析过[非流动性因子ILLIQ - 寻找市场对弱流动性的收益补偿 ](https://uqer.datayes.com/community/share/57ba8b1e228e5b79a575a226])。
