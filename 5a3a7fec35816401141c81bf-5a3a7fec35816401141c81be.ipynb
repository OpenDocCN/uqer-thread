# factor decay

```python
from quartz_extensions import *
import numpy as np
import pandas as pd


start = '2017-07-01'                       # 回测起始时间
end = '2017-12-01'                         # 回测结束时间
benchmark = 'ZZ500'                      # 策略参考标准
universe = DynamicUniverse('ZZ500')         # 证券池，支持股票和基金
refresh_rate = 'month'                     # 调仓频率，表示每周五结束之后生成调仓
delay_number = 5                         # 信号最大衰减周期
quantile_number = 5                      # 分位分组的组数
function_list = []                       # 信号处理函数列表，从前到后进行处理
industry_type = 'SW1'                    # 行业分类类型

# 因子定义
def fac_indicator(data, dependencies=['PE','PCF'], max_window=1):
    yesterday = data['PCF'].index[-1]
    PCF = 1.0 / data['PCF'].ix[-1]
    sig_PCF = pd.Series(standardize(neutralize(winsorize(PCF),yesterday)))
    PE = 1.0 / data['PE'].ix[-1]
    sig_PE = pd.Series(standardize(neutralize(winsorize(PE),yesterday)))
    CUM = sig_PCF + sig_PE
    return CUM

# 注册因子名称
def initialize(account):
    a = Signal('MY_fac', fac_indicator)
    account.signal_generator = SignalGenerator(a)

# 调用自定义因子分析函数
create_factor_tear_sheet('MY_fac', start=start, end=end, benchmark=benchmark, universe=universe, refresh_rate=refresh_rate, delay_number=delay_number, quantile_number=quantile_number, function_list=function_list, industry_type=industry_type, initialize=initialize)
```
