## 4.1.1 水泥数据

自己写的回归类，帮助理解，实现了如下功能：
- 多元回归（含截距项和不含截距项）参数$\hat{\beta}$的估计
    - $\hat{\beta} = (X'X)^{-1}X'Y$
- 参数标准差、t值、p值的计算
    - $\sigma_{\epsilon}^2 =\frac{\hat{\sigma_{\epsilon}^2}}{n-k} = \frac{\sum_{i=1}^{n} (y_i-\hat{y_i})^2}{n-k}$
    - $ t_j =\frac{\beta_j - 0}{\sigma_{\epsilon}}$
    - $ p_j = 1 - cdf(|t_j|)$
- 方差分析：SST,SSR,SSE的计算以及F值的计算
    - SST = $\sum_{i=1}^{n} (y_i - \bar{y_i})^2$
    - SSR = $\sum_{i=1}^{n} (\hat{y_i} - \bar{y_i})^2$
    - SSE = $\sum_{i=1}^{n} (y_i - \hat{y_i})^2$
    - SSE + SSR = SST
- $r^2$ 和调整后$r^2$的计算
    - $ r^2 = \frac{SSR}{SST} $
    - $ adj \\_r^2 = 1 - (1 - r^2) \frac{n-1}{n-k}$

```python
# -*- coding: utf-8 -*-

import numpy as np                          # 导入数组与矩阵计算库numpy，重新命名为np(国际惯例)
import pandas as pd                         # 导入数据框库pandas，重命名为pd(惯例)


class Regression:
    def __init__(self, X, Y):
        """
        多元线性回归
        """
        self.X = X                          # 自变量
        self.Y = Y                          # 因变量
        self.sample_num = self.X.shape[0]   # 样本数
        self.column_num = self.X.shape[1]   # 变量数
        self._beta = None                   # 参数估计值
        self._sse = None                    # 残差平方和
        self._ssr = None                    # 回归平方和
        self._sst = None                    # 总平方和
        self._std = None                    # 参数标准差
        self._t  = None                     # 参数t值
        self._R2 = None                     # r2
        self._adj_R2 = None                 # 调整后r2
        self._F = None                      # F值
        self._check()

    def _check(self):
        """
        输入参数检查
        """
        assert self.sample_num == len(self.Y), '样本数不一致'
        assert self.sample_num > self.column_num, '样本数不可以少于变量数'

    def fit(self, has_con=True, quiet=False):
        """
        回归拟合，计算beta估计值，以及检验参数
        :param has_con:     是否有截距项
        :param quiet:       是否输出结果
        """
        if has_con:
            cons_array = np.ones((self.sample_num, 1))          # 如果有截距项，则样本矩阵加上全为1的一列
            self.X = np.column_stack((cons_array, self.X))
            self.column_num += 1                                # 列数加1
        X_mat = np.matrix(self.X)                               # 将二维列表转为矩阵，方便运算
        self._temp = (X_mat.T * X_mat).I                        # 计算(X.T X)^{-1}，保存为私有属性，避免重复计算
        self._beta = self._temp * X_mat.T * self.Y              # 根据公式 beat_hat = (X.T * X)^{-1} * X.T * Y ,计算beta_hat
        self._get_std()                                         # 回归参数检验，根据t值和p值
        self._get_R()                                           # 方差分析
        if not quiet: self.summary()                            # 如果需要输出结果，则打印结果表

    def _get_std(self):
        X_mat = np.matrix(self.X)                               # 二维数组转矩阵
        Y_mat = np.matrix(self.Y)
        epsilon_mat = Y_mat - X_mat * self._beta                # 计算残差向量
        self._sse = np.sum(np.array(epsilon_mat)**2)            # 计算残差平方和
        variance_mat = self._sse / (self.sample_num - self.column_num) * self._temp     # 计算残差方差的估计结果，并且得到方差矩阵
        self._std = np.sqrt(variance_mat.diagonal())            # 计算标准差向量
        self._t = np.array(self._beta).reshape((1,self.column_num)) / self._std         # 计算t值向量
        self._get_p()                                           # 计算p值向量

    def _get_p(self):
        """
        根据t值计算p值（正态分布前提）
        """
        from scipy.stats import norm
        self._p = 1 - norm.cdf(abs(self._std))

    def _get_R(self):
        self._sst = sum((self.Y - self.Y.mean())**2)            # 计算总平方和
        self._ssr = self._sst - self._sse                       # 相减得到回归平方和
        self._R2 = self._ssr / self._sst                        # 相除得到r方
        self._adj_R2 =  1 - (1 - self._R2) * (self.sample_num - 1) / (self.sample_num - self.column_num)    # 计算调整后R2
        self._F = (self._ssr / (self.column_num - 1)) / (self._sse / (self.sample_num - self.column_num))   # 计算F值

    def summary(self):
        print("\t \t \t Analysis of Regression \t \t \t")
        print("SSR: %.2f \t SSE   : %.2f \t SST: %.2f " % (self._ssr, self._sse, self._sst))
        print("R2 : %.2f \t R2_adj: %.2f \t F  : %.2f \t" % (self._R2, self._adj_R2, self._F))
        print("----------------------------------------------------------------------------------")
        print("beta_hat:  " + "\t".join(["%08.4f" for i in range(self.column_num)]) % tuple(np.array(self._beta).reshape((1,self.column_num))[0]))
        print("std     :  " + "\t".join(["%08.4f" for i in range(self.column_num)]) % tuple(np.array(self._std)[0]))
        print("t       :  " + "\t".join(["%08.4f" for i in range(self.column_num)]) % tuple(np.array(self._t)[0]))
        print("p       :  " + "\t".join(["%08.4f" for i in range(self.column_num)]) % tuple(self._p[0]))
```

```python
df = pd.read_excel('4_1_4.xlsx')                                    # 读取excel数据
X = np.array(df.loc[:, ['x1 ','x2','x3','x4']])                     # 提取自变量
Y = np.array(df.loc[:, ['Y']])                                      # 提取因变量
lr = Regression(X, Y)
lr.fit()
```

## 例4.2.1
逐步回归法求最优自变量

```python
def stepwise_regression(df):
    """
    前XIANG逐步回归
    :param df: 数据框 dataframe
    """
    x_name_set = set(['x1 ','x2','x3','x4'])                    # 自变量名称的集合
    selected = []                                               # 挑选的自变量（挑选组）
    current_score, best_score = 0.0, 0.0                        # 初始化迭代变量
    Y = np.array(df.loc[:, ['Y']])                              # 计算因变量
    step = 0                                                    # 计数
    while x_name_set and current_score == best_score:           # 当自变量全部尝试过或者是此次增加变量反而使得效果变差时终止
        score_candidate = []                                    # 得分与候选项的元组
        for candidate in x_name_set:                            # 遍历每个元素，添加到现有的挑选组里,计算得分
            selected_temp = selected + [candidate]
            X = np.array(df.loc[:, selected_temp])
            lr = Regression(X, Y)                               # 构建回归实例
            lr.fit(quiet=True)                                  # 回归拟合
            score = lr._adj_R2                                  # 采取调整后R2作为评价变量
            score_candidate.append((score, candidate))
        score_candidate.sort()                                  # 找出当前最优的变量
        best_score, best_candidate = score_candidate.pop()
        if current_score < best_score:                          # 当前最优是全局较优解时，替换成原全局较优解
            x_name_set.remove(best_candidate)                   # 自变量名称提出改变量，不再参与迭代
            selected.append(best_candidate)                     # 挑选组添加该变量
            current_score = best_score                          # 最优解替换
            step += 1                                           # 计数加一
            print("step: %s, selected: %s, score: %s" % (step, selected, current_score))
    return selected, best_score
```

```python
stepwise_regression(df)
```

## 例4.2.2
全子集法求最优自变量

```python
def full_subset_regression(df):
    """
    全集合，暴力枚举，找最优解
    """
    import itertools                                                    # 迭代工具箱，用以生成子集
    x_name_set = set(['x1 ', 'x2', 'x3', 'x4'])                         # 自变量名称集合
    best_selected, best_score = [], 0.0                                 # 初始化
    Y = np.array(df.loc[:, ['Y']])
    for i in range(len(x_name_set)):
        for selected in itertools.combinations(x_name_set, i+1):        # 生成固定大小为i+1的子集
            X = np.array(df.loc[:, selected])                           # 子集对应的自变量矩阵
            lr = Regression(X, Y)                                       # 构建回归实例
            lr.fit(quiet=True)                                          # 回归拟合
            current_score = lr._adj_R2
            print(selected, current_score)
            if current_score > best_score:                              # 当前最优是全局较优解时，替换成原全局较优解
                best_score = current_score
                best_selected = selected
    print('------------------------------------------------------------')
    print("best_x: %s, best_adj_r2: %s" % (best_selected, best_score))
```

```python
full_subset_regression(df)
```
