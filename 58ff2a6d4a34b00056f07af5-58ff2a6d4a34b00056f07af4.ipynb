>[红利策略和其他策略的比较——沪深300与它的增强兄弟连](https://xueqiu.com/5251898839/84400353)

**指数代码**：H30364
**策略描述**：剔除沪深300指数中流动性最差100只股票，使用最小方差加权，在这种情况下，波动率越小，与其他股票越不相关的股票，权重越大。
**调仓周期**：半年
**波动率**：股票过去一年的日收益率(也就是每日涨跌幅)的标准差。
**流动性**：姑且认为周平均换手率
