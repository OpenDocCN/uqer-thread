```python
import numpy as np
import pandas as pd
import talib
from talib import func
import matplotlib
import datetime
from CAL.PyCAL import *
from __future__ import division

maxHistoryDays = 22

cal = Calendar('China.SSE')
df_stat = pd.DataFrame(columns=['AvePercent1','AvePercent3','AvePercent5','AvePercent10','Counter'])

startDate = '20120801'
endDate = '20120930'
    
#主要计算区
dateList = cal.bizDatesList(startDate,endDate)
universe = set_universe('A')
#set_universe('A')

wk_startDate = cal.advanceDate(dateList[0], '-20B', BizDayConvention.Following)
wk_endDate = cal.advanceDate(dateList[-1], '+10B', BizDayConvention.Following)
    
#取得大盘的ID，沪市，深市Or创业板
dapan_indexID = ['000001.ZICN','399001.ZICN','399006.ZICN']
    
#获得大盘涨跌幅
df_dapan = DataAPI.MktIdxdGet(tradeDate=u"",indexID=dapan_indexID,ticker=u"",beginDate=wk_startDate,endDate=wk_endDate,field=u"indexID,tradeDate,closeIndex,CHGPct",pandas="1")
    
debug_counter = 0
for s in universe: 

    debug_counter = debug_counter +1
    #取得历史数据
    df = DataAPI.MktEqudAdjGet(tradeDate=u"",secID=s,ticker=u"",beginDate=wk_startDate,endDate=wk_endDate,field=u"",pandas="1")
    df['comCounter'] = 0
        
    if len(df) <= 0:
        continue
    #虽然之前使用SecSTGet过滤ST股票，但是有些股票在买的时候还不是ST股，在回测期间变成ST股。SecSTGet不输出。所以加强甄别   
    secName = str(df[df.index==0].secShortName[0])      
    if 'ST' in secName[0]:
        continue  
    
    dapan_ID = ''
    #取得大盘的ID，沪市，深市Or创业板
    if s[0] == '6':
        dapan_ID = '000001.ZICN'
    elif s[0] == '0':
        dapan_ID = '399001.ZICN'
    elif s[0] == '3':
        dapan_ID = '399006.ZICN'   
   
    print secName,dapan_ID,debug_counter,len(universe)      
        
    #计算所有日期的值
    for date in dateList:      
 
        upDays = 0
        index = 0
        while abs(upDays) < 20:

            #转换日期格式
            strDays = '%d' %index
            strDays = strDays + 'B'  
            wk_date = cal.advanceDate(date, strDays, BizDayConvention.Following)
            
            #获得股票涨跌幅
            if len(df[df.tradeDate==str(wk_date)].closePrice) == 0:
                upDays = 0                
                break
            closePrice = df[df.tradeDate==str(wk_date)].closePrice.tolist()[0]
            openPrice = df[df.tradeDate==str(wk_date)].openPrice.tolist()[0] 
            preClosePrice = df[df.tradeDate==str(wk_date)].preClosePrice.tolist()[0]  
            
            if openPrice == 0:
                upDays = 0
                break

            stk_per =  (closePrice-preClosePrice)/preClosePrice      
            
            #获得大盘涨跌幅
            dapan_per = df_dapan[df_dapan.indexID==dapan_ID][df_dapan.tradeDate==str(wk_date)].CHGPct.tolist()[0]
            
            if stk_per > dapan_per:
                if upDays >= 0:
                    upDays = upDays + 1
                else:
                    break
            elif stk_per <= dapan_per:
                if upDays <= 0:
                    upDays = upDays - 1
                else:
                    break            
            #更新index
            index = index-1
       
        #跳过无数据的日子
        if upDays == 0:
            continue
        
        get_df_index = df[df.tradeDate==str(date)].index
        
        #设置counter     
        #df.set_value(get_df_index,'comCounter',upDays)
        
#============================================            
        
        avePer1 = 0.0
        avePer3 = 0.0
        avePer5 = 0.0
        avePer10 = 0.0
        newCounter = 0
        if upDays not in df_stat.index:
            avePer1 = 0.0
            avePer3 = 0.0
            avePer5 = 0.0
            avePer10 = 0.0            
            newCounter = 0   
        else:
            avePer1 = df_stat['AvePercent1'][upDays]
            avePer3 = df_stat['AvePercent3'][upDays]
            avePer5 = df_stat['AvePercent5'][upDays]
            avePer10 = df_stat['AvePercent10'][upDays]             
            newCounter = df_stat['Counter'][upDays]

        #获得当日股票涨幅
        tomorrow = cal.advanceDate(date, '+1B', BizDayConvention.Following)
        if len(df[df.tradeDate==str(tomorrow)].openPrice) == 0 or len(df[df.tradeDate==str(date)].openPrice) == 0:
            continue          
        t_openPrice = df[df.tradeDate==str(tomorrow)].openPrice.tolist()[0]
        openPrice = df[df.tradeDate==str(date)].openPrice.tolist()[0]       
        thisPercent1 = (t_openPrice-openPrice)/openPrice


        #获得3日股票涨幅       
        wk_date = cal.advanceDate(date,'+3B', BizDayConvention.Following) 
        if len(df[df.tradeDate==str(wk_date)].openPrice) == 0:
            continue        
        t_openPrice = df[df.tradeDate==str(wk_date)].openPrice.tolist()[0]
        thisPercent3 = (t_openPrice-openPrice)/openPrice
        
        #获得5日股票涨幅       
        wk_date = cal.advanceDate(date,'+5B', BizDayConvention.Following) 
        if len(df[df.tradeDate==str(wk_date)].openPrice) == 0:
            continue          
        t_openPrice = df[df.tradeDate==str(wk_date)].openPrice.tolist()[0]                 
        thisPercent5 = (t_openPrice-openPrice)/openPrice
                
        #获得10日股票涨幅
        wk_date = cal.advanceDate(date,'+10B', BizDayConvention.Following) 
        if len(df[df.tradeDate==str(wk_date)].openPrice) == 0:
            continue          
        t_openPrice = df[df.tradeDate==str(wk_date)].openPrice.tolist()[0]            
        thisPercent10 = (t_openPrice-openPrice)/openPrice
        
        sumPer1 = avePer1*newCounter + thisPercent1
        sumPer3 = avePer3*newCounter + thisPercent3
        sumPer5 = avePer5*newCounter + thisPercent5
        sumPer10 = avePer10*newCounter + thisPercent10
        newPercent1 = sumPer1/(newCounter+1.0)
        newPercent3 = sumPer3/(newCounter+1.0)
        newPercent5 = sumPer5/(newCounter+1.0)
        newPercent10 = sumPer10/(newCounter+1.0)
        
        df_stat.set_value(upDays,'AvePercent1',newPercent1)
        df_stat.set_value(upDays,'AvePercent3',newPercent3)
        df_stat.set_value(upDays,'AvePercent5',newPercent5)
        df_stat.set_value(upDays,'AvePercent10',newPercent10)
        df_stat.set_value(upDays,'Counter',(newCounter+1))
        
    print df_stat.sort(columns='AvePercent1',ascending=False)
```
