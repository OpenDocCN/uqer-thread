```python
# -*- coding: utf-8 -*-
# choose target stocks for LZ
# created on Apr 16,2019

# 开启缓存，当前Notebook所有DataAPI数据都会在内存进行缓存
DataAPI.settings.cache_enabled = False #调试时设为False

# 导入功能模块
import numpy as np
import pandas as pd
import CAL.PyCAL as cal
```

```python
today='20190415'  #设置查询日期
```

```python
# 条件1：总股本小于20亿
df=DataAPI.EquGet(secID=u"",ticker=u"",equTypeCD=u"A",listStatusCD=u"L",field=u"ticker,secShortName,totalShares",pandas="1")
df1=df[df['totalShares']<=2000000000]
```

```python
# 条件2：PE小于15，PB小于3
df=DataAPI.MktStockFactorsOneDayGet(tradeDate=today,ticker=u'',field=u"ticker,PE,PB",pandas="1")
df2=df[(df['PB']<=3)&(df['PE']<=15)]
```

```python
# 条件3：毛利率大于30，净资产收益率大于15
df=DataAPI.FdmtIndiRtnGet(reportType=u"A",field=u"ticker,endDate,grossMARgin,ROEA",pandas="1")
df=df.groupby(by='ticker').first().reset_index()
df3=df[(df['grossMARgin']>=30)&(df['ROEA']>=15)]
```

```python
# 条件4：经营活动产生的现金净流量大于0 
df=DataAPI.FdmtCFAllLatestGet(reportType=u"A",field=u"ticker,endDate,NCFOperateA",pandas="1")
df=df.groupby(by='ticker').first().reset_index()
df4=df[df['NCFOperateA']>=0][['ticker','NCFOperateA']]
```

```python
# 合并数据
df5=pd.merge(df1,df2,how='inner',on='ticker')
df5=pd.merge(df5,df3,how='inner',on='ticker')
df5=pd.merge(df5,df4,how='inner',on='ticker')
f=lambda x:'{:.1f}'.format(x)
df5['总股本（万股）']=df5['totalShares']/10000
df5['总股本（万股）']=df5['总股本（万股）'].apply(f)
df5['经营现金净流量（万元）']=df5['NCFOperateA']/10000
df5['经营现金净流量（万元）']=df5['经营现金净流量（万元）'].apply(f)
df5=df5[['ticker','secShortName','PE','PB','endDate','grossMARgin','ROEA','总股本（万股）','经营现金净流量（万元）']]
df5.columns=['代码','名称','市盈率','市净率','最新报告时间','毛利率%','净资产收益率%','总股本（万股）','经营现金净流量（万元）']
df5.to_csv('lz.csv')
df5
```
