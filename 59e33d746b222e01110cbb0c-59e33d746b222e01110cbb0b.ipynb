# 事件类-高管持股策略

```python
import numpy as np
import pandas as pd
from CAL.PyCAL import *    # CAL.PyCAL中包含font
from datetime import timedelta,datetime
from string import zfill
```

```python
raw_df = pd.read_csv('executives_holds_all_new.csv',encoding = 'gbk')
raw_df.index = pd.to_datetime(raw_df['date'])
raw_df['ticker'] = raw_df['ticker'].apply(str).apply(lambda x:zfill(x,6))  #右对齐，填充为6个数字
raw_df1=raw_df[raw_df['num']<=10]
raw_df2= raw_df1[raw_df1['change_money']>3000000 ].sort_values(by='change_money')   #变动金额大于300W
raw_df2 =raw_df2[raw_df2['duty']=='executive']  #增减持类型为高管
raw_df2 =raw_df2[raw_df2['direction']=='zengchi']   #筛选出增持的事件
```
