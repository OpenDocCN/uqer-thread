# 

**Alpha #18**
$$(-1 * rank(((stddev(abs((close - open)), 5) + (close - open)) + correlation(close, open,
10))))$$

先解读该公式：<br>
1.计算前10天的收盘价于开盘价的价差；<br>
2.计算前5天价差绝对值的标准差，并与前一天价差相加；<br>
3.对2的结果进行rank操作，（关于rank操作的具体细节参考我的上一篇文章[Zura论文附录中WorldQuant的Alpha#03](https://uqer.datayes.com/community/share/5750f4a1228e5b86a61fd7dc)）<br>
4.计算过去10天收盘价与开盘价的相关系数，并返回其与3的结果的和。<br>

以下是我个人对该式子的理解：<br>
该公式主要涉及的数据是开盘价和收盘价，且有很多close-open的操作。close-open，变相代表着日内的收益，也反映了涨跌幅。观察式子我们可以发现：<br>
1.对式子的左半部分，由于系数-1的存在，close-open越小，且5天价差的标准差越小时结果越大；<br>
2.对式子的右半部分，close和open正相关且相关性越好，结果越大。<br>
因此，我们最后对close和open正相关性好的、close-open小的、5天价差的标准差小的股票做多且给予较大权重；反之自行理解。<br>

1.对于close与open正相关，我们可以想像相关系数为1，则在k线图上，open和close连成两条各段斜率都相同的线，说明股价近10天持续上升。对于股价持续上升的原因主要有：<br>
(1)价格优势支持股价上涨。经过一段时间的调整，很多股票处于超跌状态，有着强烈的反弹需求，同时，价格也相当的低，那么在这种情况下，股价有上涨的动力，而且主力资金也往往青睐这类股票，只要介入资金，往往会有一波很好的行情。<br>
(2)业绩增长支持股价上涨。业绩优良、市盈率低的股票往往受到市场主流资金的关注，往往这些股票都是白马股票，虽然不会一下子涨的很猛，但不知不觉中就已涨上去了，所以，一些有业绩支持的个股往往具有很好的上涨动力。<br>
(3)市场热点支持股价上涨。热点永远是当前股市的核心。如果，近期股市某一个板块上涨的厉害，涨停板前三十名中，同类型板块据多，那么这类板块就是热点，是很容易吸引资金和股民的关注的，这类股票涨的厉害。<br>
(4)增量资金支持股价上涨。目前，股市里的行情还是靠资金推动起来的，如果，增量资金介入的程度深，流入的资金量多，那么，这类股票上涨的程度是很大的。<br>
因此，对于这类股票买入获利的概率较大；

2.当open与close相关性不明显的时候，说明股价持续振荡，则当上个交易日的close-open越小，最好是下跌的时候，由于持续震荡我们认为这时股价上涨的概率略大；而对于价差的标准差，我们希望它比较小，说明股价的震荡幅度较小，降低了部分风险。
