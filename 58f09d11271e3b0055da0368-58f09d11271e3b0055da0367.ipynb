# 筹码分布WINNER10%获利盘买入后的走势

熟悉通达信公式的都知道，在通达信里有winner函数， 获利盘比例.
用法:WINNER(CLOSE),表示以当前收市价卖出的获利盘比例,例如返回0.1表示10%获利盘;WINNER(10.5)表示10.5元价格的获利盘比例
本人一直认为这是一个有意义的函数，winner的计算就是依据换手率，从以前的筹码分布依次移出、增加新筹码数量，得当新一日的筹码分布，股价的震荡在筹码分布里可以明显看到波峰和波谷，波峰和波谷也可以理解为股价的压力位和支撑位。
那么使用winner函数是否真的有用，下面用量化程序说话。
本文没有以模拟盘的形式说明winner函数的效果，换个思路：以当日收盘价为计算价格，收盘价在低于10%获利盘的进行买入10支(随机，没有排序)，每隔20交易日取一买入一次，共执行13次，记录这13 * 10=130支股票（实际少于130支，因为有的日期没有筛选到10支低于10%获利盘的票），买入后120个交易日内的股价变化。

```python
#选股后若干天，股价的走势导出计算 
import pandas as pd
from pandas import DataFrame
import numpy as np
import datetime 
from CAL.PyCAL import *
cal = Calendar('China.SSE')
def get_stockchipwin(date): 
    chipwin10 = pd.read_csv('upload/outputChippdT100101.csv')[['secID',date]]
    closePrice = DataAPI.MktEqudAdjGet(tradeDate=date,isOpen="",field=u"secID,closePrice",pandas="1")
    chipwin10.set_index('secID', inplace=True)
    closePrice.set_index('secID', inplace=True)
    newset = pd.merge(chipwin10, closePrice, how='outer', left_index=True, right_index=True)
    newset=newset[newset[date]>0]
    return newset[newset[date]>=newset['closePrice']][date].head(10)

idate='2015-08-25'
allsecID = pd.DataFrame()
for k in range(1,13):
    bdate = cal.advanceDate(idate,str(k*20)+'B',BizDayConvention.Preceding)
    bdate = bdate.strftime('%Y-%m-%d')
    secID = pd.DataFrame(get_stockchipwin(bdate))  
    for i in range(1,120):
        begindate = cal.advanceDate(bdate,str(i)+'B',BizDayConvention.Preceding)  
        begindate = begindate.strftime('%Y-%m-%d')
        closePrice = DataAPI.MktEqudAdjGet(tradeDate=begindate,secID=list(secID.index),isOpen="",field=u"secID,closePrice",pandas="1")
        closePrice.columns=['secID',str(i)]
        closePrice.set_index('secID', inplace=True)
        if len(closePrice)>0:
            secID[str(i)] = closePrice[str(i)]
            secID[str(i)] = [x/y for x,y in zip(secID[str(i)],secID[bdate]) ]
    secID = secID.drop([bdate],axis=1)
    allsecID = allsecID.append(secID)
allsave = pd.DataFrame()
for i in range(1,120):
    allsave[str(i)] = list(allsecID.sort_values([str(i)], ascending=False)[str(i)])
allsave.to_csv('chipwin10.csv')
print 'ok'
```
