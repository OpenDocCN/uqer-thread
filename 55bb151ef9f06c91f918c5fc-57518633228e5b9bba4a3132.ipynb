-  于是，又想进行尝试，每次触发止盈或者止损后，在随后的一段时间不操作（称之为冷却时间cool_day）,看看效果如何

#####  发现结果变差了，为什么会产生这种情况呢？？
#####  直接去bt里寻找答案，就看最近这段时间的行情变化（2015年6月18日--2015年7月30日）

```python
bt.tail(30)
```


-  在开始大跌期间，2015年7月8日之前，策略的信号(delta)一直都是negative，躲过了初始的这一波大跌

-  从7月13日就开始给出买入信号（delta为正），可以看到在过去两天大盘已经反弹了11%，从这一点也可以很直观的感觉到均线策略的滞后性（在趋势性行情非常好用，但是在震荡行情则经常会反其道而行之）

-  在接下来的小幅反弹中，信号一直为negative，终于在24日有了微弱的买入信号，然后在28号发生了止损操作

-  在发生止损后，随后的cool_day是不进行操作的，不过29号的signal也为负，但是今天却发生了小幅反弹

##### 从上述分析可以看到，策略的好坏稳健等因素还是要在signal方面做文章，止盈止损可以是增强策略的一个方式，但也并非每个策略加入止盈止损后效果会有提升

##### 以我个人的实践经验来看，盘中止损的效果可能会更好一些，由于上面的策略是日度策略，止损操作其实是对止损后一日的涨跌进行判断，但是对于日度策略，好的止损应该做到盘中止损，比如下午2点冲破了止损线，那就应该清仓；若是对于周度策略，可以以日收盘价来进行止损操作，以此类推
