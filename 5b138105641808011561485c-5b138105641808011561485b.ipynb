# 第三步 进行数据处理导入优矿
- 然后数据处理整合成一个dataframe导入优矿。这个就比较简单，直接上代码。最后生成的df是纵坐标是股票代码，横坐标是时间，内容是扣非净利润。

```python
import pandas  as pd
import  numpy as np
import pymysql

# 打开数据库连接
db = pymysql.connect("localhost", "root", "123456", "test")

# 使用 cursor() 方法创建一个游标对象 cursor
cursor = db.cursor()

sql = "SHOW TABLES FROM `test`"

# 执行SQL语句
cursor.execute(sql)
# 获取所有记录列表
results = cursor.fetchall()
secli = []
df = pd.DataFrame()
print (len(results))
for i in results:
    sql = "select * from `%s`"%(i)
    cursor.execute(sql)
    res = cursor.fetchall()

    tmp1 = []
    tmp2 = []
    if res != ():
        secli.append("sec"+i[0])
        for q  in res:
            tmp1.append(q[0])
            tmp2.append(q[1])
        tmp1 = np.array(tmp1)
        df1 = pd.DataFrame(tmp2)
        df1 = df1.T
        df1.columns = (tmp1)
        df = pd.concat((df,df1),axis=0)
secli = np.array(secli)

df = df.set_index(secli)
df.to_csv('final.csv')
print (df)
```

# 第四步 多进程简易优化爬取速度（可选）
- 像scrapy那种爬虫框架里面多进程什么的都给你写好了，但如果用传统方法爬取数据肯定很慢，就像刚才的爬取操作花了609秒。所以如果我们要爬3000只股票呢，或者更多更多，那速度根本太慢，于是需要用多进程进行优化。由于Python有GIL锁存在，所以多线程基本可以说没啥用。但是异步单进程是有效的，主要用于网页返回response那里节省速度。这里不使用异步单进程举例，简单使用多进程进行优化。
- 网上教程太多又杂，什么代码都有。比如爬取网页加头文件 加cookie urllib能写2种以上写法，直接不用记那些操作，引入requests就相当方便。 多进程也一样，实际上最实用的就是那个pool。我们只需要写个函数，这个函数所需要的参数是迭代的，那么我们就可以把这个函数和这个迭代的参数一起扔进pool里面。然后就能自动使用多进程进行运算了。
- 于是我们这个思路就是写个函数，然后拿股票进行迭代，每个股票的爬取写数据库操作就是一个进程。正好我们前面写了那个for循环。直接删了for 然后拿个函数包上。在下面写pool的实例就是默认用全部进程，加上if \__name\__ 那段。这个涉及系统方面操作不写这个会报错。下面放代码。 

```python
import requests
import multiprocessing as mp
import re
import pymysql
import time
# 打开数据库连接
db = pymysql.connect("localhost", "root", "123456", "test")

# 使用 cursor() 方法创建一个游标对象 cursor
cursor = db.cursor()

t1 = time.time()

# 使用预处理语句创建表
ti = []
secdata  = open('C:/Users/yehuo/PycharmProjects/financial/sec.txt','r',encoding='utf-8')
for h in secdata.readlines():
    h.encode('utf-8').decode('utf-8-sig')
    h=h.replace('.XSHG','')
    h=h.replace('.XSHE','')
    h=h.replace('\n','')
    ti.append(h)
if ti[-1]=="":
    ti.pop(-1)
print (ti)

def caozuo(ti):
    ticker = ti
    # 使用 execute() 方法执行 SQL，如果表存在则删除
    cursor.execute("DROP TABLE IF EXISTS `%s`"%(ticker))
    sql = """CREATE TABLE `%s` (DATETIME  CHAR(20) ,KFINCOME CHAR(20) )"""%(ticker)
    cursor.execute(sql)

    url = 'http://quotes.money.163.com/f10/zycwzb_%s.html#01c01'%(ticker)
    print (url)
    response = requests.get(url)
    data = response.text
    data = str.encode(data)
    data = str(data,encoding='utf-8')
 

    #获取数据
  
    pattern = re.compile('<table class="table_bg001 border_box limit_sale scr_table" >.*</table>',re.S)
    result = pattern.findall(data)


    # fh2.write(result[0])
    # fh2.close()

    #筛选时间 插入数据
    line = result[0].splitlines()
    pattern = re.compile('<tr class="dbrow"> <th >(.*)</th> </tr>')
    res = pattern.findall(line[1])
    if res != []:
        res = res[0]
        res = res.split('</th><th >')
        print(len(res),res[0],res[1],res[-1],res[-2])
        pattern = re.compile('<tr>(.*)</td></tr>')
        res2 = pattern.findall(line[12])
        res2 = res2[0]
        res2 = res2.replace("</td><td class='cRed'>","</td><td>")
        res2 = res2.replace("<td class='cRed'>","")
        res2 = res2.split('</td><td>')
        res2[0]=res2[0].replace('<td>','')
        print(len(res2),res2[0],res2[1],res2[-1],res2[-2])
        final = zip(res,res2)
        for i in final:
            # SQL 插入语句
            sql = """INSERT INTO `%s`(DATETIME,KFINCOME)VALUES ('%s','%s')"""%(ticker,i[0],i[1])
            cursor.execute(sql)
                # 提交到数据库执行
            db.commit()
    return  0

def multi():
    pool = mp.Pool()
    pool.map(caozuo,ti)
if __name__ == '__main__':
    multi()
    t2 = time.time()
    print (t2-t1)
```
