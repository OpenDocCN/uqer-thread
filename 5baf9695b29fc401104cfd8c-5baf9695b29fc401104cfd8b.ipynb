# look 1 look HMM模式识别

### <font color=#0099ff>**基于HMM模式识别模型的股市走势预测**</font>：

---
1. 首先，按照事先分类，选取历史上属于同类走势的日期以及该日期之前若干个星期的股票数据，提取股票数据中某些特征指标（成交价格，成交量，等等）形成相应的序列作为模型的输入，并应用Baum-Welch算法对各类模型进行训练

2. 其次，根据训练好的HMM模型，选取若干个星期的股票特征指标（成交价格，成交量，等等）序列作为输入，应用向前－向后算法计算各个模型发生的概率，选取最大概率对应的模型，从而得到下一阶段股票走势的识别结果


### <font color=#0099ff>**HMM策略实证结果**</font>
---
择时指数：沪深300
时间区间：2016/01/01 至 2018/09/27

观测序列变量如下：

- X1: 股票日收益率；
- X2: 资金日净流入占当日成交金额的比例；
- X3: 成交金额日环比；
- X4: 标准化成交金额，即：（日成交金额－过去一年平均成交金额）/ 过去一年成交金额波动率。
- X5: 沪股通日净买入占当日沪股通成交金额的比例；

运用观测变量组合每期滚动训练HMM模型，每期的训练样本长度为 1 年（由于沪股通的数据较少）

并根据沪深300指数的涨跌情况，将所有样本划分为六类：
- （熊市上涨，熊市下跌，震荡上涨，震荡下跌，牛市上涨，牛市下跌）

```python
import pandas as pd
import numpy as np
from pandas import Series,DataFrame,concat
from scipy import interpolate
import math
import time
from hmmlearn.hmm import GaussianHMM
from scipy.interpolate import InterpolatedUnivariateSpline
from CAL.PyCAL import *
cal = Calendar('China.SSE')
```

```python
def HMM(ticker='000300',beginDate='2017-01-01',endDate='2018-01-01'):
    
    last50d = cal.advanceDate(beginDate, '-50B').strftime('%Y-%m-%d')
    cal_dates = DataAPI.TradeCalGet(exchangeCD=u"XSHG", beginDate=last50d, endDate=endDate).sort('calendarDate')
    cal_dates = cal_dates[cal_dates['isOpen'] == 1]
    date_list = cal_dates['calendarDate'].values.tolist()
    date_list = [x.replace("-", "") for x in date_list]
    
    #date = Date.parseISO(beginDate)
    #begin = cal.advanceDate(date, '-50B', BizDayConvention.Preceding)
    #hist_VIX = getHistDailyVIX(begin, Date.parseISO(endDate))
    #VIX_skew = hist_VIX.VIX.rolling(20).apply(lambda x: skew_(x))
    #VIX_skew = np.array(VIX_skew[-len(CHGPct):])
    #VIX = np.array(hist_VIX.VIX[-len(CHGPct):])
    
    last1y = cal.advanceDate(beginDate, '-1Y').strftime('%Y-%m-%d')
    CHGPct = DataAPI.MktIdxdGet(ticker=ticker,beginDate=beginDate,endDate=endDate,
                                field=['tradeDate','CHGPct'],pandas="1")['CHGPct']
    netInflow = DataAPI.MktIdxMfGet(ticker='000001',beginDate=beginDate,endDate=endDate,
                                  field=['tradeDate','netInflow'],pandas="1")['netInflow']
    TurnoverValue = DataAPI.MktIdxdGet(ticker=ticker,beginDate=last1y,endDate=endDate,
                                     field=['tradeDate','turnoverValue'],pandas="1")['turnoverValue']
    CHGPct = np.array(CHGPct)
    nf_pct = np.array(netInflow) / np.array(TurnoverValue[-len(CHGPct):])
    tv_diff = np.array(np.diff(TurnoverValue)[-len(CHGPct):])
    tv_std = np.array(pd.rolling_std(TurnoverValue, window=200)[-len(CHGPct):])
    tv_mean = np.array(pd.rolling_mean(TurnoverValue, window=200)[-len(CHGPct):])
    tv_stand = (np.array(TurnoverValue[-len(netInflow):]) - tv_mean) / tv_std
    
    hk = pd.DataFrame()
    for date in date_list:
        temp = DataAPI.equSHHKstatsTDGet(tradeDate=date,exchangeCD=u"SHSC",field=['tradeDate','buyValue','totalValue'],pandas="1")
        temp['b_pct'] = temp['buyValue'] / temp['totalValue']
        hk = pd.concat([hk, temp],axis=0)
    hk_b = np.array(hk['b_pct'][-len(CHGPct):])
    
    X = np.column_stack([CHGPct,nf_pct,tv_diff,tv_stand,hk_b])
    model = GaussianHMM(n_components=6, covariance_type="diag", n_iter=5000).fit([X])
    hidden_states = model.predict(X)
    
    res = pd.DataFrame({'CHGPct':CHGPct,'state':hidden_states})
    for i in range(model.n_components):
		state = (hidden_states==i)
		idx = np.append(0,state[:-1])
		res['CHGPct%s'%i] = res.CHGPct.multiply(idx,axis=0)
		res[str(i)] = (res['CHGPct%s'%i] +1).cumsum() -1   
    state_ = res[['0','1','2','3','4','5']]
    state_ = state_.tail(1).iloc[0]
    state_ = state_.sort_values(ascending=False,inplace=False)
    state = 0
    if int(hidden_states[-1]) == int(state_.index[0]) or int(hidden_states[-1]) == int(state_.index[1]):
        state = 1
    elif int(hidden_states[-1]) == int(state_.index[5]):
        state = -1
    return state	
```
