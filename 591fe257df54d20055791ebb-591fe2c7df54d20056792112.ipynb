接上贴--[fama多因素模型](https://uqer.datayes.com/community/share/58ecd011e4ebdb005667d065)

barra多因子模型构建因子收益率和因子载荷系统预测未来股票投资收益率和波动风险。
因关于因子载荷和因子收益率的设定有不同概念，这里主要采用华泰研报内容，故标称为华泰多因子。   
华泰多因子将计算的因子数值经过去极值标准化之后定为因子暴露度，   
然后将多元回归的自变量系数认为是因子收益率，通过面板数据分期进行截面回归得到自变量   
权重系数矩阵是因子收益率，通过使用时间序列方法预测因子收益率，根据新一期因子暴露度   
计算未来收益率，并根据因子收益率的历史波动情况预估新一期预测的波动率，并通过一系列   
的统计和数值优化的方式将投资组合波动率降低。

回归法是一种最常用的测试因子有效性的方法，具体做法是将因子在第T期的暴露度与T+1期的    
股票收益进行线性回归，所得到的回归系数即为因子在T期的因子收益率，同时还能得到该   
因子收益率在本期回归中的显著度水平--t 值。   
我们的回归模型为:
$r_i^{T+1} = \sum_j X_j^Tf_{jt}^T + X_d^Td_i^T + \mu_i^T$
$r_i^{T+1}$:股票 i 在第 T + 1 期的收益率 
$d_i^T$:股票 i 在在第 T 期因子 d 上的暴露度 
$f_{jt}^T$:股票 i 在第 T 期第 j 个行业因子上的暴露度(属于该行业则为1，否则为0) 
$X_j^T (X_d^T)$:第 T 期第 j 个行业因子(或因子d)的因子收益率，需回归拟合 
$\mu_i^T$:股票 i 在第 T 期的残差收益率

在所有截面期上，我们对因子d进行回归测试，能够得到该因子的因子收益率序列（即所有截   
面期回归系数$X_d^T$构成的序列）和对应的t值序列。t值指的是对单个回归系数$X_d^T$   
的t检验统计量，描述的是单个变量显著性，t值的绝对值大于临界值说明该变量是显著的,   
即该解释变量（T期个股在因子d的暴露度）是真正影响因变量（T+1期个股收益率）的一   
个因素。也就是说，在每个截面期上，对于每个因子的回归方程，我们设<br />
假设检验$H_0:X_d^T=0$
备择假设$H_0:X_d^T\not=0$

该假设检验对应的t统计量为
$t=\frac{X_d^T}{SE(X_d^T)}$
其中$SE(X_d^T)$代表回归系数$X_d^T$的标准差的无偏估计量。 一般t值绝对值大于2我们就认为本
期回归系数$X_d^T$是显著异于零的 （也就是说,本期因子d对下期收益率具有显著的解释作用）。
注意，我们在回归模型中加入了行业因素，能在一定程度上规避行业因素对成长因子的影
响。并且回归法单因子测试中并未对市值因素的影响进行调整，但是在后续计算因子IC值的
部分会做一些调整。 

测试模型构建方法如下： 
1． 股票池：全 A 股，剔除 ST 股票，剔除每个截面期下一交易日停牌的股票。 
2． 回溯区间：2011-01-01 至 2016-12-31。 
3． 截面期：每个自然月的最后一个交易日计算因子暴露度，与下一整个自然月
的个股超额收益（以沪深 300 指数为基准）进行回归。 
4． 数据处理方法： 
b) [中位数去极值](https://uqer.datayes.com/community/share/58e5b36eaf30b600580729d9)
c) 标准化：将去极值处理后的因子暴露度序列减去其现在的均值、除以其标准差，
得到一个新的近似服从N(0,1)分布的序列，这样做可以让不同因子的暴露度之间
具有可比性； 
d) 缺失值处理：dropna()

5． 回归权重：由于普通最小二乘回归（OLS）可能会夸大小盘股的影响（因为小盘股的估
值等因子出现极端值概率较大，且小盘股数目很多，但占全市场的交易量比重较小），
并且回归可能存在异方差性， 故我们参考 Barra 手册， 采用加权最小二乘回归 （WLS） ，
使用个股流通市值的平方根作为权重，此举也有利于消除异方差性

6． 因子评价方法： 
a) t 值序列绝对值平均值——因子显著性的重要判据； 
b) t 值序列绝对值大于 2 的占比——判断因子的显著性是否稳定； 
c) 因子收益率序列平均值，以及该均值零假设检验的 t 值——判断因子收益率序列是
否方向一致，并且显著不为零； 
d) t 值序列均值的绝对值除以 t 值序列的标准差——结合显著性和波动性， 辅助判断因
子是否有效、稳健。

```python
from __future__ import division
import numpy as np
import pandas as pd
import matplotlib.pylab as plt
import seaborn as sns
import statsmodels.api as sm

from functools import wraps
def wrapper_func_cache(func):
    @wraps(func)
    def wrapped(*args, **kwargs):
        if not hasattr(func, '__cache_result'):
            func.__cache_result = {}
        
        hash_key = [str(i) for i in args] + [str(j) for j in kwargs.iteritems()]
        cache_key = hash(tuple(hash_key))
        
        if cache_key not in func.__cache_result:
            func.__cache_result[cache_key] = func(*args, **kwargs)
            
        return func.__cache_result[cache_key]
    return wrapped

@wrapper_func_cache
def Monthly_const_trade_calendar(beginDate=u"20050101",endDate=u""):
    """交易日历，可以选择cache，只调用 api 一次
    """
    data = DataAPI.TradeCalGet(exchangeCD=u"XSHG,XSHE", beginDate=beginDate, endDate=endDate, field=['calendarDate', 'isMonthEnd'])
    data = data[data.isMonthEnd == 1].calendarDate.tolist()
    data = sorted(list(set(data))) 
    return data

# ST过滤
def st_remove(source_universe, st_date=None):
    """
    给定股票列表,去除其中在某日被标为ST的股票
    Args:
        source_universe (list of str): 需要进行筛选的股票列表
        st_date (datetime): 进行筛选的日期,默认为调用当天
    Returns:
        list: 去掉ST股票之后的股票列表

    Examples:
        >> universe = set_universe('A')
        >> universe_without_st = st_remove(universe)
    """
    st_date = st_date if st_date is not None else datetime.datetime.now().strftime('%Y%m%d')
    df_ST = DataAPI.SecSTGet(secID=source_universe, beginDate=st_date, endDate=st_date, field=['secID'])
    return [s for s in source_universe if s not in list(df_ST['secID'])]

# bug 函数封装问题
def winsorize_median(tmp):
    if isinstance(tmp, pd.Series):
        mc = sm.stats.stattools.medcouple(tmp)
        data = sorted(tmp)
        q1 = np.percentile(data, 25)
        q3 = np.percentile(data, 75)
        iqr = q3-q1
        if mc >= 0:
            l = q1-1.5*np.exp(-3.5*mc)*iqr
            u = q3+1.5*np.exp(4*mc)*iqr
        else:
            l = q1-1.5*np.exp(-4*mc)*iqr
            u = q3+1.5*np.exp(3.5*mc)*iqr
        tmp[tmp<l]=l
        tmp[tmp>u]=u
    return tmp   

# bug 函数封装问题
def standardize_zscore(tmp):
    if isinstance(tmp, pd.Series):
        mu = tmp.mean()
        sigma = tmp.std()
        tmp = (tmp - mu)/sigma
    return tmp   

# 获取申万一级行业分类成份股编码 
def Stock_Industry(point):
    # 申万一级指数编码
    industrys=DataAPI.IndustryGet(industryVersion=u"SW",industryVersionCD=u"",industryLevel=u"1",isNew=u"1",field=u"",pandas="1")
    indexes=list(industrys['indexSymbol'].astype(str))#行业对应的指数编码
    index_param=DataAPI.IdxGet(secID=u"",ticker=indexes,field=u"ticker,secShortName,publishDate",pandas="1")#获取指数上市时间即为对应行业开始时间
    #2014年前已存在的行业
    old_industry=index_param[index_param['publishDate']<'2014-01-01']
    old_indexes=list(old_industry['ticker'])
    old_industry_name=list(old_industry['secShortName'])
    #2014年后的所有行业
    new_industry=index_param
    new_indexes=list(new_industry['ticker'])
    new_industry_name=list(new_industry['secShortName'])
    # 申万一级指数分类函数
    SW_Old = old_indexes
    SW_New = new_indexes    
    if isinstance(point,str):
        point = pd.to_datetime(point)
    timepoint = pd.to_datetime('2014-01-01')
    if point <= timepoint:
        return SW_Old
    elif point > timepoint:
        return SW_New
    else:
        print 'wrong'

# 获取申万一级行业分类成份股       
def Industry_Constituent_Stocks(industry, timepoint):
    tmp = DataAPI.IdxConsGet(ticker=industry, intoDate=timepoint, isNew=u"",field=['consID'],pandas="1")
    return sorted(list(tmp.consID.values))
```

```python
# 获取交易日历
trade_calendar = Monthly_const_trade_calendar(beginDate=u"20110101",endDate=u"20170510")

# temp
date = '2017-03-31'
fac = 'PB'

# 获取当日股票池
univ = set_universe('A',date)
univ = st_remove(univ, date)

# 获取横截面股票收益率
date_next = trade_calendar[trade_calendar.index(date)+1]
Price_front = DataAPI.MktEqudGet(tradeDate=date, secID=univ, beginDate=u"",endDate=u"",isOpen="1",field=['secID','closePrice','accumAdjFactor','negMarketValue'],pandas="1")
Price_front['front'] = Price_front.closePrice / Price_front.accumAdjFactor
Price_behind = DataAPI.MktEqudGet(tradeDate=date_next, secID=univ, beginDate=u"",endDate=u"",isOpen="",field=['secID','closePrice','accumAdjFactor'],pandas="1")
Price_behind['behind'] = Price_behind.closePrice / Price_behind.accumAdjFactor
R_T = pd.merge(Price_front.loc[:,['secID','front','negMarketValue']], Price_behind.loc[:,['secID','behind']], on='secID')
R_T['W'] = np.sqrt(R_T.negMarketValue)
R_T['return'] = np.log(R_T.behind/R_T.front)

# 获取横截面因子载荷
factor_loadings = DataAPI.MktStockFactorsOneDayGet(tradeDate=date, secID=univ, field=['secID']+[fac],pandas="1")
factor_loadings[fac] = 1.0/factor_loadings[fac]
factor_loadings[fac] = standardize_zscore(winsorize_median(factor_loadings[fac]))
#factor_loadings[fac].plot()
factor_loadings.set_index('secID',inplace=True, drop=False)
#print factor_loadings.describe()


# 获取行业暴露度 哑变量矩阵
SW_Industry_List = Stock_Industry(date)
Linear_Regression = pd.DataFrame()
for i in SW_Industry_List:
    i_Constituent_Stocks = Industry_Constituent_Stocks(i, date)
    i_Constituent_Stocks = list(set(i_Constituent_Stocks).intersection(set(univ)))
    try:
        tmp = factor_loadings.loc[i_Constituent_Stocks]
        tmp.dropna(inplace=True)
        tmp[i]=1.0
    except:
        print i
    Linear_Regression = Linear_Regression.append(tmp)
Linear_Regression.fillna(0.0, inplace=True)

# HS300指数收益率
R_index = DataAPI.MktIdxmGet(beginDate=u"20110101",endDate=u"20170510",indexID=u"000300.ZICN",ticker=u"",field=['endDate','closePrice'],pandas="1")
R_index.sort_values('endDate',inplace=True)
R_index['R_index'] = np.log(R_index.closePrice.shift(-1)/R_index.closePrice)
R_index.set_index('endDate', inplace=True)

Linear_Regression = pd.merge(Linear_Regression, R_T.loc[:,['secID','return','W']], on='secID')
Linear_Regression.set_index('secID', inplace=True)

X = Linear_Regression.loc[:,SW_Industry_List+[fac]]
y = Linear_Regression['return'] - R_index.loc[date,'R_index']

# OLS回归
est=sm.OLS(y,X)
result = est.fit()
print result.summary()

# WLS回归
wls = sm.WLS(y, X, weights=Linear_Regression.W)
wls = wls.fit()
print wls.summary()
```

## 因子 IC 值的计算 
因子的 IC 值是指因子在第 T 期的暴露度与 T+1 期的股票收益的相关系数，即 
$IC_d^T = corr(R^{T+1}, d^T)$
其中，$IC_d^T$代表因子d在第T期的IC值，$R^{T+1}$代表所有个股第T+1期的收益率向量，
$d^T$表所有个股第 T 期在因子 d 上的暴露度向量。
测试模型构建方法如下： 
1. 股票池、回溯区间、截面期均与回归法相同。 
2. 因子值在去极值、标准化、去空值处理后，在截面期上用其做因变量对市值因子及行业   
因子（哑变量）做线性回归，取残差作为因子值的一个替代。这样做可以消除行业因素    
和市值因素对因子的影响。 计算残差序列和 T+1 期股票收益序列的相关系数作为 T 期因   
子 IC 值。 
3. 因子评价方法：    
a) IC 值序列的均值大小——因子显著性；    
b) IC 值序列的标准差——因子稳定性；     
c) IR 比率（IC 值序列均值与标准差的比值）——因子有效性；     
d) IC 值累积曲线——随时间变化效果是否稳定；     
e) IC 值序列大于零的占比——因子作用方向是否稳定。

## 分层模型回测 
- 分层模型主要是考量保序作用。    


依照因子值对股票进行打分，构建投资组合回测，是最直观的衡量指标优劣的手段。一般来
说，通过回归法和计算因子 IC 值都无法确定因子的单调性（例如，某因子值排名在中间 1/3
的个股表现比前 1/3 和后 1/3 的个股表现要好） ， 但是分层回测法是可以确定因子单调性的。
分层回测法逻辑简单，结果清晰，操作方便，并且具有能区分因子单调性的独特优势，是接
受度非常高的一种单因子测试手段。

```python
from __future__ import division
import numpy as np
import pandas as pd

import datetime

def st_remove(source_universe, st_date=None):
    """
    给定股票列表,去除其中在某日被标为ST的股票
    Args:
        source_universe (list of str): 需要进行筛选的股票列表
        st_date (datetime): 进行筛选的日期,默认为调用当天
    Returns:
        list: 去掉ST股票之后的股票列表

    Examples:
        >> universe = set_universe('A')
        >> universe_without_st = st_remove(universe)
    """
    st_date = st_date if st_date is not None else datetime.datetime.now().strftime('%Y%m%d')
    df_ST = DataAPI.SecSTGet(secID=source_universe, beginDate=st_date, endDate=st_date, field=['secID'])
    return [s for s in source_universe if s not in list(df_ST['secID'])]
```

```python
# 申万一级指数编码
industrys=DataAPI.IndustryGet(industryVersion=u"SW",industryVersionCD=u"",industryLevel=u"1",isNew=u"1",field=u"",pandas="1")
indexes=list(industrys['indexSymbol'].astype(str))#行业对应的指数编码
index_param=DataAPI.IdxGet(secID=u"",ticker=indexes,field=u"ticker,secShortName,publishDate",pandas="1")#获取指数上市时间即为对应行业开始时间
#2014年前已存在的行业
old_industry=index_param[index_param['publishDate']<'2014-01-01']
old_indexes=list(old_industry['ticker'])
old_industry_name=list(old_industry['secShortName'])
#2014年后的所有行业
new_industry=index_param
new_indexes=list(new_industry['ticker'])
new_industry_name=list(new_industry['secShortName'])
# 申万一级指数分类函数
SW_Old = old_indexes
SW_New = new_indexes
def Stock_Industry(point):
    timepoint = pd.to_datetime('2014-01-01')
    if point <= timepoint:
        return SW_Old
    elif point > timepoint:
        return SW_New
    else:
        print 'wrong'
# 获取申万一级行业分类成份股       
def Industry_Constituent_Stocks(industry, timepoint):
    tmp = DataAPI.IdxConsGet(ticker=industry, intoDate=timepoint, isNew=u"",field=['consID'],pandas="1")
    return sorted(list(tmp.consID.values))
```

## [五分位测试](https://uqer.datayes.com/community/share/578f04e0228e5b3b9b5f1ab7)

```python
fac = 'PB'

# 可编辑部分与 strategy 模式一样，其余部分按本例代码编写即可

# -----------回测参数部分开始，可编辑------------
start = '2011-01-01'                       # 回测起始时间
end = '2017-01-01'                         # 回测结束时间
benchmark = 'HS300'                        # 策略参考标准
universe = DynamicUniverse('A')  # 证券池，支持股票和基金
capital_base = 1e9   
freq = 'd'
refresh_rate = Monthly(-1)  
# ---------------回测参数部分结束----------------
# 把回测参数封装到 SimulationParameters 中，供 quick_backtest 使用
sim_params = quartz.SimulationParameters(start, end, benchmark, universe, capital_base)
# 获取回测行情数据
idxmap, data = quartz.get_daily_data(sim_params)
# 运行结果
results = {}
# 调整参数(选取股票的Q因子五分位数)，进行快速回测
for quantile_five in range(0, 5):
    
    # ---------------策略逻辑部分----------------
    refresh_rate = Monthly(-1)  
    commission = Commission(0.0002, 0.0002)
    
    def initialize(account):                   # 初始化虚拟账户状态
        pass
    
    def handle_data(account):               

        Operation_Date = account.current_date.strftime('%Y%m%d') 

        # 过滤ST股票
        univ = account.universe
        univ = st_remove(univ, st_date=Operation_Date)
        #print len(univ)

        # 判断行业分类
        Industrys = Stock_Industry(account.current_date)

        # 获取行业划分成份股
        Constituent_Stocks = {}
        for s in Industrys:
            tmp = Industry_Constituent_Stocks(s, Operation_Date)
            tmp = list(set(tmp).intersection(set(univ)))
            Constituent_Stocks[s] = tmp

        # 获取因子数据
        factor_ = DataAPI.MktEqudGet(tradeDate=account.previous_date.strftime('%Y%m%d'),secID=univ, isOpen="", field=['secID']+[fac], pandas="1")
        factor_.dropna(inplace=True)
        factor_.set_index('secID', inplace=True)

        # 按行业排序筛选
        holding_list = []
        holding = {}
        for s in Industrys:
            tmp = Constituent_Stocks[s]
            tmp = factor_.loc[tmp,:]
            tmp.sort_values(fac, inplace=True)
            q_min = tmp.quantile(quantile_five*0.2)[fac]
            q_max = tmp.quantile((quantile_five+1)*0.2)[fac]
            industry_univ = tmp[tmp[fac]>=q_min][tmp[fac]<=q_max]
            industry_univ = industry_univ.index.values.tolist()
            holding[s] = industry_univ
            holding_list = holding_list + industry_univ

        # 伪 行业中性
        neu_weights = {}
        neu = set_universe('HS300',account.previous_date.strftime('%Y%m%d')) + set_universe('ZZ500',account.previous_date.strftime('%Y%m%d'))
        tmp = DataAPI.MktEqudGet(tradeDate=account.previous_date.strftime('%Y%m%d'),secID=neu, isOpen="",field=['secID','negMarketValue'],pandas="1")
        tmp.set_index('secID', inplace=True)
        for s in Industrys:
            cm = list(set(neu).intersection(set(holding[s])))
            cm = tmp.loc[cm]
            cm = cm.negMarketValue.sum()
            neu_weights[s] = cm
        neu_sum = pd.Series(neu_weights).sum()

        # 仓位操作
        # 不在股票池中的，清仓
        for stk in account.security_position:
            if stk not in holding_list:
                order_to(stk, 0)
        # 在目标股票池中的，等权买入
        for s in Industrys:
            neu_weight = neu_weights[s]/neu_sum
            order_univ = holding[s]
            for i in order_univ:
                order_pct_to(i, 1.0/len(order_univ)*neu_weight)
# ---------------策略逻辑部分结束----------------

    # 把回测逻辑封装到 TradingStrategy 中，供 quick_backtest 使用
    strategy = quartz.TradingStrategy(initialize, handle_data)
    # 回测部分
    bt, acct = quartz.quick_backtest(sim_params, strategy, idxmap, data, refresh_rate=refresh_rate, commission=commission)

    # 对于回测的结果，可以通过 perf_parse 函数计算风险指标
    perf = quartz.perf_parse(bt, acct)

    # 保存运行结果
    tmp = {}
    tmp['bt'] = bt
    tmp['annualized_return'] = perf['annualized_return']
    tmp['volatility'] = perf['volatility']
    tmp['max_drawdown'] = perf['max_drawdown']
    tmp['alpha'] = perf['alpha']
    tmp['beta'] = perf['beta']
    tmp['sharpe'] = perf['sharpe']
    tmp['information_ratio'] = perf['information_ratio']
    
    results[quantile_five] = tmp
    print str(quantile_five)
print 'done'   
```

```python
import matplotlib.pyplot as plt
from matplotlib import rc
rc('mathtext', default='regular')
import seaborn as sns
sns.set_style('white')
from matplotlib import dates
import time
import scipy.stats as st
from CAL.PyCAL import font    # CAL.PyCAL中包含font


fig = plt.figure(figsize=(14,9))
fig.set_tight_layout(True)
ax1 = fig.add_subplot(211)
ax2 = fig.add_subplot(212)
ax1.grid()
ax2.grid()

for qt in results:
    bt = results[qt]['bt']

    data = bt[[u'tradeDate',u'portfolio_value',u'benchmark_return']]
    data['portfolio_return'] = data.portfolio_value/data.portfolio_value.shift(1) - 1.0   # 总头寸每日回报率
    data['portfolio_return'].ix[0] = data['portfolio_value'].ix[0]/	1e9 - 1.0
    data['excess_return'] = data.portfolio_return - data.benchmark_return                 # 总头寸每日超额回报率
    data['excess'] = data.excess_return + 1.0
    data['excess'] = data.excess.cumprod()                # 总头寸对冲指数后的净值序列
    data['portfolio'] = data.portfolio_return + 1.0     
    data['portfolio'] = data.portfolio.cumprod()          # 总头寸不对冲时的净值序列
    data['benchmark'] = data.benchmark_return + 1.0
    data['benchmark'] = data.benchmark.cumprod()          # benchmark的净值序列
    results[qt]['hedged_max_drawdown'] = max([1 - v/max(1, max(data['excess'][:i+1])) for i,v in enumerate(data['excess'])])  # 对冲后净值最大回撤
    results[qt]['hedged_volatility'] = np.std(data['excess_return'])*np.sqrt(252)
    results[qt]['hedged_annualized_return'] = (data['excess'].values[-1])**(252.0/len(data['excess'])) - 1.0
    # data[['portfolio','benchmark','excess']].plot(figsize=(12,8))
    # ax.plot(data[['portfolio','benchmark','excess']], label=str(qt))
    ax1.plot(data['tradeDate'], data[['portfolio']], label=str(qt))
    ax2.plot(data['tradeDate'], data[['excess']], label=str(qt))
    

ax1.legend(loc=0)
ax2.legend(loc=0)
ax1.set_ylabel(u"净值", fontproperties=font, fontsize=16)
ax2.set_ylabel(u"对冲净值", fontproperties=font, fontsize=16)
```
