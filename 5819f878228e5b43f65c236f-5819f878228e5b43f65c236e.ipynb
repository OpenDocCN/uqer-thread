一：基于社区里原有的帖子 “再议隐马尔可夫模型的市场择时策略” 的再探讨
1：简介：
    （1）：N^2的参数矩阵（转移矩阵），要初始化一个[1,。。。]_1*n的向量(可以认为第一个状态是1) 这样就有初始的状态，初始的概率转移矩阵，以及默认给出的系宗里的n个状态  
    （2）：实际中我们并不清楚是什么状态，尽管我们有此时此刻的数据（因此有一个明显的数据集，和一个隐藏的状态） 状态序列和隐藏的状态序列是概率相关的。因此将这种类型的过程建模为有一个隐藏的马尔科夫过程（因为隐藏的状态是转化的，是有概率转移矩阵的）和一个与这个隐藏马尔科夫过程 概率相关 的并且可以观察到的（准确的数据集）状态集合，就是隐马尔可夫模型。 
    （3）：含有隐含未知参数的马尔可夫过程，核心在于：从可观察的参数中（提供的数据集）确定该过程的隐含参数 举个例子：三个骰子，（6个面 4个面，8个面）随机取一个投，得到一个确定的序列（可见状态）（数据集2321546412378，但是有一个隐藏的状态D6D4D8D8D4D6...这些状态之间存在着转移概率，那么这些概率我们可以自己设定，比如D6后面不可以是D6，D8后面是D4的概率为0.9...）（D是隐含状态） 因此实际观察状态和隐含转态之间有一个输出概率（元素：可见 隐含 输出概率 隐含状态转移概率）隐含转态转换图可以画出 
    （4）：缺失的数据：
	算法主要分为三类，分别解决三种问题： 知道骰子有几种（隐含状态数量），每种骰子是什么（转换概率），根据掷骰子掷出的结果（可见状态链），我想知道每次掷出来的都是哪种骰子（隐含状态链）。 知道骰子有几种（隐含状态数量），每种骰子是什么（转换概率），根据掷骰子掷出的结果（可见状态链），我想知道掷出这个结果的概率。 知道骰子有几种（隐含状态数量），不知道每种骰子是什么（转换概率），观测到很多次掷骰子的结果（可见状态链），我想反推出每种骰子是什么（转换概率）。

```python
from hmmlearn.hmm import GaussianHMM
import numpy as np
from matplotlib import cm, pyplot as plt
import matplotlib.dates as dates
import pandas as pd
import datetime
```

```python
n = 3 #6个隐藏状态
start = '2012-01-01'                       # 回测起始时间
end = '2016-10-21'                         # 回测结束时间
ticker_name='000001'
data_sz=DataAPI.MktIdxdGet(ticker=ticker_name,beginDate=start,endDate=end,field=u"",pandas="1")
data=data_sz[['tradeDate','preCloseIndex','openIndex','lowestIndex','highestIndex','closeIndex','turnoverVol','turnoverValue']]
print data[0:5]
volume=data['turnoverVol']
close=data['closeIndex']
close2=data['preCloseIndex']
logDel = np.log(np.array(data['highestIndex'])) - np.log(np.array(data['lowestIndex']))
logRet_1 = np.array(np.diff(np.log(close2)))                         #这个作为后面计算收益使用
logRet_5 = np.log(np.array(close[5:])) - np.log(np.array(close[:-5]))#5日指数对数收益差
logVol_5 = np.log(np.array(volume[5:])) - np.log(np.array(volume[:-5]))
logDel = logDel[5:]
logRet_1 = logRet_1[4:]
close = close[5:]
Date = pd.to_datetime(data['tradeDate'][5:])
A = np.column_stack([logDel,logRet_5,logVol_5])#3个特征 理解成3维数据
print A[0:2]                                   #格式注意
```

关于 covariance_type 的参数有下面四种：
（1）：spherical：是指在每个马尔可夫隐含状态下，可观察态向量的所有特性分量使用相同的方差值。对应协方差矩阵的非对角为0，对角值相等，即球面特性。这是最简单的高斯分布PDF。 
	特点：spherical是最简单的，通常用在数据不足或者硬件平台性能有限的情况之下
（2）：image diag：是指在每个马尔可夫隐含状态下，可观察态向量使用对角协方差矩阵。对应协方差矩阵非对角为0，对角值不相等。
	特点：diag是hmmlearn里面的默认类型。 
（3）：image full：是指在每个马尔可夫隐含状态下，可观察态向量使用完全协方差矩阵。对应的协方差矩阵里面的元素都是不为零。 
	特点：full是最强大的，但是需要足够多的数据来做合理的参数估计
（4）：image tied：是指所有的马尔可夫隐含状态使用相同的完全协方差矩阵。
	特点：tied则可以看作是GaussianHMM和GMMHMM的特有实现
这四种PDF类型里面，spherical, diag和full代表三种不同的高斯分布概率密度函数

选自知乎“Python hmmlearn中的混淆矩阵是怎么表示的？”

```python
model = GaussianHMM(n_components= n, covariance_type="full", n_iter=2000).fit([A])
hidden_states = model.predict(A)
hidden_states[0:10]
```

一些问题
1：以上的是获得全部的数据训练然后展示，因此结果仅仅是观察分类的正确性不具有预测意义（多数情况下训练的结果是很准的）
2：下面进行样本内训练，估计预测样本外的数据，再根据效果图评价好坏。
3:以什么样的形式来评价预估的好坏呢？ 根据累积的收益曲线（针对不同的隐含状态）（如果线性区分性很明显说明预测的很准确）
4：对以下两个累计收益曲线来看：样本内的效果显而易见很好嘛 样本外紫线下跌状态分类的比较明确 绿色上涨状态也比较明确，但是红色的状态不是很明确

实验一：选取样本内 2011-01-01 to 2014-12-31（刚好跳过15年的时间段）
		样本外2015-01-01 to 2016-04-30

```python
n=5#6个隐藏状态
start = '2011-01-01'                       # 回测起始时间
end = '2014-12-31'                         # 回测结束时间
ticker_name='000001'
data_sz=DataAPI.MktIdxdGet(ticker=ticker_name,beginDate=start,endDate=end,field=u"",pandas="1")
data=data_sz[['tradeDate','preCloseIndex','openIndex','lowestIndex','highestIndex','closeIndex','turnoverVol','turnoverValue']]
#print data[0:5]
volume=data['turnoverVol']
close=data['closeIndex']
close2=data['preCloseIndex']
logDel = np.log(np.array(data['highestIndex'])) - np.log(np.array(data['lowestIndex']))
logRet_1 = np.array(np.diff(np.log(close2)))                         #这个作为后面计算收益使用
logRet_5 = np.log(np.array(close[5:])) - np.log(np.array(close[:-5]))#5日指数对数收益差
logVol_5 = np.log(np.array(volume[5:])) - np.log(np.array(volume[:-5]))
logDel = logDel[5:]
logRet_1 = logRet_1[4:]
close = close[5:]
Date = pd.to_datetime(data['tradeDate'][5:])
A = np.column_stack([logDel,logRet_5,logVol_5])#3个特征 理解成3维数据
#print A[0:2]                                   #格式注意
model = GaussianHMM(n_components= n, covariance_type="full", n_iter=2000).fit([A])
hidden_states = model.predict(A)
hidden_states[0:10]
plt.figure(figsize=(14, 6)) 
print model
for i in range(model.n_components):
    pos = (hidden_states==i)
    plt.plot_date(Date[pos],close[pos],'o',label='hidden state %d'%i,lw=3)
    plt.legend(loc="left")
res = pd.DataFrame({'Date':Date,'logRet_1':logRet_1,'state':hidden_states}).set_index('Date')
plt.figure(figsize=(15, 6)) 
for i in range(model.n_components):
    pos = (hidden_states==i)
    pos = np.append(0,pos[:-1])#第二天进行买入操作
    df = res.logRet_1 
    res['state_ret%s'%i] = df.multiply(pos)
    plt.plot_date(Date,np.exp(res['state_ret%s'%i].cumsum()),'-',label='hidden state %d'%i)
    plt.legend(loc="left")
```
