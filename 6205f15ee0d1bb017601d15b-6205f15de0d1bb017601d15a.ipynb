```python
import datetime
import numpy as np
import pandas as pd
import scipy.stats as ss
import matplotlib.pylab as plt

def rank_INT(series, c=3.0/8, stochastic=False):
    """ Perform rank-based inverse normal transformation on pandas series.
        If stochastic is True ties are given rank randomly, otherwise ties will
        share the same value. NaN values are ignored.
        Args:
            param1 (pandas.Series):   Series of values to transform
            param2 (Optional[float]): Constand parameter (Bloms constant)
            param3 (Optional[bool]):  Whether to randomise rank of ties
        
        Returns:
            pandas.Series
    """

    # Check input
    assert(isinstance(series, pd.Series))
    assert(isinstance(c, float))
    assert(isinstance(stochastic, bool))

    # Set seed
    np.random.seed(123)

    # Take original series indexes
    orig_idx = series.index

    # Drop NaNs
    series = series.loc[~pd.isnull(series)]

    rank = series.rank()

    # Convert rank to normal distribution
    transformed = rank.apply(rank_to_normal, c=c, n=len(rank))
    
    return transformed[orig_idx]

def rank_to_normal(rank, c, n):
    # Standard quantile function
    x = (rank - c) / (n - 2*c + 1)
    return ss.norm.ppf(x)
```

```python
def analysis_pct(pct):
    nav = (1+pct).cumprod()
    maxdrawdown = min((nav - nav.cummax()).div(nav.cummax()).fillna(0))
    annualized_ret = (nav.iloc[-1]-1)/(len(nav)/252.0)
    return '%.2f%%'%(annualized_ret*100), '%.2f'%((annualized_ret-0.03)/(pct.std()*252**0.5)), '%.2f%%'%(-maxdrawdown*100)
```

```python
def calc_pe_roe(stock_codes, date):
    b = (datetime.datetime.strptime(date, '%Y-%m-%d') - datetime.timedelta(days=180)).strftime('%Y%m%d')
    df_roe = DataAPI.FdmtIndiRtnPitGet(ticker=stock_codes,publishDateEnd=date,publishDateBegin=b,field=u"ticker,ROE",pandas="1")
    df_roe=df_roe.drop_duplicates('ticker', keep='first').set_index('ticker')
    df_factors = DataAPI.MktStockFactorsOneDayGet(tradeDate=date,ticker=stock_codes,field=u"ticker,PB",pandas="1")
    df_factors.set_index('ticker', inplace=True)
    df_factors = df_factors.merge(df_roe, left_index=True, right_index=True, how='inner')
    df_factors = df_factors[(df_factors>0).all(axis=1)]
    stock_codes = df_factors.index.tolist()
    y = np.log(df_factors.PB.values)
    X = rank_INT(df_factors.ROE).values
    X = np.column_stack([np.ones(len(X)), X])
    R = np.linalg.lstsq(X, y)
    residual = y-X.dot(R[0])
    factors = pd.Series(residual,index=df_factors.index)
    factors = neutralize_factors(factors.dropna(), date)
    return factors

def calc_roe(stock_codes, date):
    b = (datetime.datetime.strptime(date, '%Y-%m-%d') - datetime.timedelta(days=180)).strftime('%Y%m%d')
    df_roe = DataAPI.FdmtIndiRtnPitGet(ticker=stock_codes,publishDateEnd=date,publishDateBegin=b,field=u"ticker,ROE",pandas="1")
    df_roe=df_roe.drop_duplicates('ticker', keep='first').set_index('ticker')
    factors=df_roe.ROE.dropna()
    factors=neutralize_factors(factors, date)
    return factors

def calc_D2A_rate(stock_codes, date):
    b = (datetime.datetime.strptime(date, '%Y-%m-%d') - datetime.timedelta(days=180)).strftime('%Y%m%d')
    df_d2a_rate = DataAPI.FdmtIndiLqdGet(ticker=stock_codes,endDate=date,beginDate=b,field=u"ticker,asseTLiabRatio",pandas="1")
    df_d2a_rate=df_d2a_rate.drop_duplicates('ticker', keep='first').set_index('ticker')
    factors=df_d2a_rate.asseTLiabRatio.dropna()
    factors=neutralize_factors(factors, date)
    return factors
```

```python
def neutralize_factors(factors, date):
    """
    行业中性化、市值中性化
    """
    df_stock_industry=DataAPI.EquIndustryGet(
        ticker=factors.index.tolist(),intoDate=date,industryVersionCD=u"010303",field=u"ticker,industryName1").set_index('ticker')
    X_dummy = pd.get_dummies(df_stock_industry.industryName1)
    df_factors = DataAPI.MktStockFactorsOneDayGet(tradeDate=date,ticker=factors.index.tolist(),field=u"ticker,LCAP",pandas="1")
    df_factors.set_index('ticker', inplace=True)
    X = pd.merge(df_factors, X_dummy, left_index=True, right_index=True).dropna()
    idxs = X.index
    y = factors.loc[idxs].values
    X = X.values
    X = np.column_stack([np.ones(len(X)), X])
    R = np.linalg.lstsq(X, y)
    residual = y-X.dot(R[0])
    return pd.Series(residual, index=idxs)

def standardize(series,n=3):
    median = series.quantile(0.5)
    new_median = ((series - median).abs()).quantile(0.50)
    max_range = median + n*new_median
    min_range = median - n*new_median
    series = np.clip(series,min_range,max_range)
    return (series-series.mean())/series.std()
```

```python
def calc_sue(stock_codes, date):
    b = (datetime.datetime.strptime(date, '%Y-%m-%d') - datetime.timedelta(days=4*365)).strftime('%Y-%m-%d')
    df_netincomes = DataAPI.FdmtISQPITGet(
        ticker=stock_codes,publishDateEnd=date,beginDate=b,field='ticker,endDate,NIncome').drop_duplicates(['ticker','endDate'],keep='first')
    df_netincomes = df_netincomes.pivot_table('NIncome', index='endDate', columns='ticker')
    df_netincomes.sort_index(ascending=False, inplace=True)
    def calc_sue_impl(net_incomes):
        idx=net_incomes.fillna(method='ffill').dropna().index[0]
        net_incomes = net_incomes.loc[idx:]
        try:
            diff = net_incomes.diff(-4).iloc[1:9].dropna()
            assert len(diff) >= 4
            eps = diff.mean()
            sigma = diff.std()
            factor = (net_incomes.iloc[0] - net_incomes.iloc[4] - eps)/sigma
        except:
            factor = None
        return factor
    factors = df_netincomes.apply(calc_sue_impl, axis=0).dropna()
    factors = neutralize_factors(factors, date)
    return factors
    
```

```python
def calc_deltaroe(stock_codes, date):
    b = (datetime.datetime.strptime(date, '%Y-%m-%d') - datetime.timedelta(days=365*2)).strftime('%Y-%m-%d')
    df_roe = DataAPI.FdmtIndiRtnPitGet(ticker=stock_codes,publishDateEnd=date,publishDateBegin=b,field=u"ticker,endDate,ROE",pandas="1")
    df_roe = df_roe.pivot_table('ROE',index='endDate',columns='ticker')
    df = pd.concat([df_roe.iloc[-1]-df_roe.iloc[-5],df_roe.iloc[-2]-df_roe.iloc[-6]], axis=1)
    factors = df.apply(lambda x: x.iloc[0] if not pd.isnull(x.iloc[0]) else x.iloc[1], axis=1).dropna()
    factors = neutralize_factors(factors, date)
    return factors
```

```python
def calc_jor(stock_codes, date):
    b = (datetime.datetime.strptime(date, '%Y-%m-%d') - datetime.timedelta(days=180)).strftime('%Y-%m-%d')
    d1 = DataAPI.FdmtEeGet(ticker=stock_codes,publishDateBegin=b,publishDateEnd=date,field=u"ticker,publishDate") #业绩快报
    d1['publishDate'] = d1['publishDate'].apply(lambda x: x.split(' ')[0]) 
    d2 = DataAPI.FdmtEfNewGet(ticker=stock_codes,publishDateBegin=b,publishDateEnd=date,field=u"ticker,publishDate")#业绩预告
    d3=DataAPI.FdmtBSGet(ticker=stock_codes,publishDateEnd=date,beginDate=b,field='ticker,publishDate') #定期公告
    df_pubDate = pd.concat([d1,d2,d3])
    df_pubDate = df_pubDate.sort_values('publishDate', ascending=False).drop_duplicates('ticker', keep='first')
    df_pubDate.index=range(len(df_pubDate))
    df1 = DataAPI.MktEqudAdjGet(ticker=stock_codes,beginDate=df_pubDate['publishDate'].min(),endDate=date,field=u"ticker,tradeDate,preClosePrice,lowestPrice")
    df_jors = df1.pivot_table('lowestPrice',index='tradeDate',columns='ticker')/df1.pivot_table('preClosePrice',index='tradeDate',columns='ticker')
    df_jors = df_jors.replace({0:None}).sub(df_index_price.loc[df_jors.index,'lowestIndex']/df_index_price.loc[df_jors.index,'preCloseIndex'], axis=0)
    factors = df_pubDate.apply(lambda x: df_jors[df_jors.index>=[x.loc['publishDate']]][x.loc['ticker']].iloc[0],axis=1)
    factors.index  = df_pubDate['ticker'].values
    factors = factors.dropna()
    factors = neutralize_factors(factors, date)
    return factors
     
```

```python
def calc_gpoa(stock_codes, date):
    b = (datetime.datetime.strptime(date, '%Y-%m-%d') - datetime.timedelta(days=365*2)).strftime('%Y-%m-%d')
    df = DataAPI.FdmtISQPITGet(
        ticker=stock_codes,publishDateEnd=date,beginDate=b,field='ticker,endDate,revenue,COGS').drop_duplicates(['ticker','endDate'],keep='first')
    df['profit'] = df['revenue'] - df['COGS']
    df_profit = df.pivot_table('profit', index='endDate', columns='ticker')
    df = DataAPI.FdmtBSGet(ticker=stock_codes,publishDateEnd=date,publishDateBegin=b,reportType='Q1,S1,Q3,A',field=u"ticker,endDate,TAssets")
    df_asset = df.pivot_table('TAssets', index='endDate', columns='ticker')
    df_gpoa = df_profit/df_asset
    df = pd.concat([df_gpoa.iloc[-1],df_gpoa.iloc[-2]], axis=1)
    factors = df.apply(lambda x: x.iloc[0] if not pd.isnull(x.iloc[0]) else x.iloc[1], axis=1).dropna()
    factors = neutralize_factors(factors, date)
    return factors
    
```

```python
position = pd.Series()
n_groups = 5
position_universe = {i:pd.Series() for i in range(n_groups)}
position_chg = False
init_month = 0
tmp_prices = pd.DataFrame()
tmp_prices_univeres = {i:pd.DataFrame() for i in range(n_groups)}
df_navs = pd.DataFrame(columns=['benchmark', 'strategy'])
df_navs = pd.DataFrame(columns=['benchmark']+list(range(n_groups)))
for i, date in enumerate(df_index_price['tradeDate'][:-1]):
    if date <= beginDate:
        continue
    month_num = int(date.split('-')[1])
    if month_num != init_month:
        position_chg = True
        init_month = month_num
    if position_chg: #每月初换仓
        print(date)
        last_date = df_index_price['tradeDate'].iloc[i-1]
        eff_date = df_index_details['effDate'][df_index_details['effDate']<=last_date].iloc[-1]
        stock_codes = df_index_details[df_index_details['effDate']==eff_date]['consTickerSymbol'].sort_values().tolist()
        factors = []
        # factors.append(calc_pe_roe(stock_codes, last_date).dropna()) #添加pe_roe因子
        # factors.append(calc_roe(stock_codes, last_date).dropna()) #添加roe因子
        # factors.append(calc_sue(stock_codes, last_date).dropna()) #添加sue因子
        # factors.append(calc_jor(stock_codes, last_date).dropna()) #添加jor因子
        # factors.append(calc_deltaroe(stock_codes, last_date).dropna()) #添加delta_roe因子
        # factors.append(calc_gpoa(stock_codes, last_date).dropna()) #添加gpoa因子
        factors.append(calc_D2A_rate(stock_codes, last_date).dropna()) #添加资产负债率因子
        agg_f = pd.concat(factors, axis=1, join='inner')
        agg_f = agg_f.apply(standardize, axis=0).sum(axis=1) #因子标准化之后相加
        n = len(agg_f)//n_groups
        for i in range(n_groups):
            top_codes = agg_f.sort_values().iloc[i*n:i*n+n].index.tolist() if i<n_groups-1 else agg_f.sort_values().iloc[i*n:].index.tolist()
            et = datetime.datetime.strptime(date, '%Y-%m-%d') + datetime.timedelta(days=60)
            et = et.strftime('%Y-%m-%d')
            tmp_prices=DataAPI.MktEqudAdjGet(ticker=top_codes,beginDate=date,endDate=et,field=u"ticker,tradeDate,closePrice,preClosePrice") 
            tmp_prices = tmp_prices.pivot_table('closePrice', index='tradeDate', columns='ticker')/tmp_prices.pivot_table('preClosePrice', index='tradeDate', columns='ticker')
            weight = df_index_details[df_index_details['effDate']==eff_date]
            weight = weight[weight['consTickerSymbol'].isin(top_codes)][['consTickerSymbol', 'weight']]
            weight = pd.Series(weight['weight'].values, index=weight['consTickerSymbol'])
            weight = weight/weight.sum()
            position = position_universe[i]
            if position.empty:
                position = weight
            else:
                position = weight * position.sum()
            position_universe[i] = position
            tmp_prices_univeres[i] = tmp_prices
        position_chg = False
    df_navs.loc[date, 'benchmark'] = df_index_price.loc[date, 'closeIndex']
    for i in range(n_groups):
        tmp_prices = tmp_prices_univeres[i]
        df_pcts = tmp_prices.loc[date]
        position = position_universe[i]
        position = position * df_pcts    
        assert not pd.isnull(position.sum())
        df_navs.loc[date, i] = position.sum()
        position_universe[i] = position
df_navs = df_navs.pct_change().fillna(0)
for i in range(n_groups):
    df_navs['alpha'+str(i)] = df_navs[i] - df_navs['benchmark']
df_navs['top-bottom'] = df_navs[n_groups-1] - df_navs[0]
df_navs = (1+df_navs).cumprod()
        
        
        
    
    
    
    
    
```
