```python
#2.1numpy的属性

import numpy as np
array=np.array([[1,2,3],
               [2,3,4]])
print (array)
print'number of dim:',array.ndim#几维数组
print'shape:',array.shape#行多少列多少，就是shape
print'size:',array.size#总共有多少元素在里边，就是size

#2.2numpy的创建array
    #本节内容为numpy如何定义array的数字类型和表达形式-几行几列
import numpy as np

a=np.array([2,3,4],dtype=np.int)#int整数，后面不写数字及时默认64位
print(a.dtype)
a=np.array([2,3,4],dtype=np.int32)#numpy中有各种各样的数据形式，自行百度
print(a.dtype)
a=np.array([2,3,4],dtype=np.float32)#数字越小，精度越高
print(a.dtype)

a=np.array([[2,23,4],
            [2,32,4]])
print(a)

a=np.zeros((3,4))#生成3行4列全部为0的矩阵
print(a)

a=np.ones((3,4),dtype=np.int16)#生成3行4列全部为1的矩阵，并定义元素格式为整数16个字符
print(a)

a=np.empty((3,4))#生成3行4列为空的矩阵，实际输出接近于零
print(a)

a=np.arange(10,20,2)#arange生成有序数列a（起始值，终止值，步长）
print(a)

a=np.arange(12).reshape((3,4))#定义一个有序数列，个数为12，重新定义它的长和宽：3行4列，没有给出起始和终止则默认0-11
print(a)

a=np.linspace(1,10,5)#定义一个线段a（起始值，终止值，几段）
print(a)
a=np.linspace(1,10,6).reshape((2,3))#定义一个线段a（起始值，终止值，几段）.定义长款（2行3列）
print(a)

#2.3numpy的基础运算
    #数组的加减乘除
import numpy as np
a=np.array([10,20,30,40])
b=np.arange(4)
print(a)
print(b)
print(a,b)
c=a+b
print(c)#numpy中数组相加减就是元素逐个相加减
d=a*b
print(d)#numpy中数组相乘就是元素逐个相乘
e=b**2#numpy中数组的次方，符号为**
print(e)

    #数组的三角函数运算sin，cos，tan，cot
a=np.array([10,20,30,40])
b=np.arange(4)
c=10*np.sin(a)
print("sin:",c)

    #数组元素的比较大小
a=np.array([10,20,30,40])
b=np.arange(4)
print(b)
print"判断b中元素是否小于3："
print(b<3)#判断b中元素是否小于3，是返回true，否返回false
print"判断b中元素是否等于3："
print(b==3)#判断b中元素是否等于3，注意判断符号为==

    #矩阵运算
import numpy as np
a=np.array([[1,1],
            [0,1]])
b=np.arange(4).reshape((2,2))
print(a)
print(b)
c=a*b#元素逐个相乘
c_dot=np.dot(a,b)#矩阵相乘
c_dot_2=a.dot(b)#矩阵相乘，与上一种相乘结果一致
print(c)
print(c_dot)
print(c_dot_2)

    #[0,1)随机生成数字
import numpy as np
a=np.random.random((2,4))#第一个random是模块，第二个random是函数,随机生成2行4列的矩阵，每个元素大小在（0,1）之间
print(a)
print(np.sum(a))#求和
print(np.max(a))#寻找做大最小值
print(np.min(a))

    #矩阵内部求和和寻找最大最小值
import numpy as np
a=np.random.random((2,4))
print(a)
print("axis=1每一行求和：")
print(np.sum(a,axis=1))#axis=1每一行求和
print("axis=1每一行的最大值：")
print(np.max(a,axis=1))#axis=1每一行最大值
print("axis=1每一列的最大值：")
print(np.max(a,axis=0))#axis=1每一列最大值

#2.4numpy的基础运算2
import numpy as np
A=np.arange(2,14).reshape((3,4))#定义一个序列2-13，3行4列
print(A)
print("找到A中最小值的索引")
print(np.argmin(A))#找到A中最小值的索引即位置
print("找到A中最大值的索引")
print(np.argmax(A))#找到A中最大值的索引即位置
print("找到A中的平均值")#下面两种都是均值
print(np.mean(A))
print(A.mean())
print("找到A中的中位数")
print(np.median(A))
print("累加A中元素：")
print(np.cumsum(A))
print("累差A中元素：")
print(np.diff(A))
print("找出A中非零的数：")#输出的结果是两个数列，第一个数列代表非零数所在行，第二个数列代表非零数所在列
print(np.nonzero(A))
print("矩阵转置")
print(np.transpose(A))
print(A.T)
print((A.T).dot(A))#矩阵相乘，不知道哪个乘以哪个？？？

print("逐行排序：")#不是所有元素排序，而是每一行从小到大排序
A=np.arange(14,2,-1).reshape((3,4))
print(A)
print(np.sort(A))
print("截取修改clip功能：")
print(A)
print(np.clip(A,5,9))#A中的所有元素，所有小于5的元素都变成5，所以大于9的元素都变成9,5-9之间的元素不变
print("该列平均值axis=0：")
print(np.mean(A,axis=0))
print("该行平均值axis=1：")
print(np.mean(A,axis=1))




```
