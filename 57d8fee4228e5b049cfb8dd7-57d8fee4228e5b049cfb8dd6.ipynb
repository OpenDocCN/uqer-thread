```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from CAL.PyCAL import *
```

```python
#得出股票类分级的代码、名称
name=DataAPI.FundGet(category='E',operationMode=u"",field=u"secID,ticker,exchangeCd,tradeAbbrName,isClass",pandas="1")
name=name.loc[(name['isClass']==2) & ((name['exchangeCd'] == 'XSHE') | (name['exchangeCd'] == 'XSHG'))].sort('exchangeCd')
name.head()
```

```python
#生成通用的SecID
ticker0=name['secID'].tolist() 
```

```python
#得出不同利率
rates=DataAPI.FundLeverRateGet(secID=ticker0,beginDate=u"",endDate=u"",field=u"beginDate,ticker,returnRateA,note",pandas="1").sort('beginDate',ascending=False).drop_duplicates(['ticker'])
rates=rates[['ticker','returnRateA','note']]
rates.replace(['一年期定期存款利率(税后)+3.0%','一年期定期存款利率(税后)+3%','一年期定期基准利率(税后)+3%','一年期存款基准利率(税后)+3%','一年期定期存款利率(税后)+3.00%','一年期银行定期存款利率(税后)+3%'],'+3.0%',inplace=True)
rates.replace(['一年期定期存款利率(税后)+3.5%','一年期定期存款利率+3.5%','一年期同期银行定期存款利率+3.5%'],'+3.5%',inplace=True)
rates.replace(['一年期定期存款利率(税后)+3.2%'],'+3.2%',inplace=True)
rates.replace(['一年期定期存款利率(税后)+4%','一年期定期存款利率(税后)+4.0%','一年期定期存款利率(税后)+4.00%'],'+4.0%',inplace=True)
rates.replace(['一年期定期存款利率(税后)+4.5%'],'+4.5%',inplace=True)
rates.replace(['一年期定期存款利率(税后)+5%'],'+5.0%',inplace=True)
rates.head()
```

```python
#获取基金价格
price=DataAPI.MktFunddGet(secID=ticker0,beginDate=u"20150101",endDate=u"",field='ticker,tradeDate,closePrice,circulationShares',pandas="1")
price.head()
```

```python
#获取基金净值
nav=DataAPI.FundNavGet(secID=ticker0,beginDate=u"20150101",field='ticker,endDate,NAV',pandas="1")
nav.columns=['ticker','tradeDate','NAV']
nav.head()
```

```python
df=pd.merge(price,nav)
```

```python
#得出隐含收益率yield（约），此为基础表格
df=pd.merge(df,rates)
df['yield']=df['returnRateA']/(df['closePrice']-df['NAV']+1)/100
```

```python
#将不同的分级基金分组
df30=df[df['note']=='+3.0%']
df32=df[df['note']=='+3.2%']
df35=df[df['note']=='+3.5%']
df40=df[df['note']=='+4.0%']
df45=df[df['note']=='+4.5%']
df50=df[df['note']=='+5.0%']
```

```python
#将均价、平均收益率、总份额放到一个表中，默认用df30，如果需要其他的利率，直接修改第一行的数字即可
grouped=df40.groupby('tradeDate')
get_wavg1=lambda g: np.average(g['closePrice'],weights=g['circulationShares'])
get_wavg2=lambda g: np.average(g['yield'],weights=g['circulationShares'])
get_wavg3=lambda g: np.sum(g['circulationShares'])
price=grouped.apply(get_wavg1)
Yield=grouped.apply(get_wavg2)
shares=grouped.apply(get_wavg3)/100000000
classA=pd.DataFrame({'均价':price,'平均收益率':Yield,'总份额（亿）':shares})
classA=classA.drop('2016-04-29')
#绘图
#plt.figure(1)
ax1=plt.subplot(311)
ax2=plt.subplot(312)
ax3=plt.subplot(313)
plt.sca(ax1)
classA['均价'].plot(figsize=(14,9))
plt.title('分级A均价'.decode('utf-8'), fontproperties = font)

plt.sca(ax2)
classA['平均收益率'].plot()
plt.title('隐含收益率'.decode('utf-8'), fontproperties = font)

plt.sca(ax3)
classA['总份额（亿）'].plot()
plt.title('总份额（亿）'.decode('utf-8'), fontproperties = font)
```
