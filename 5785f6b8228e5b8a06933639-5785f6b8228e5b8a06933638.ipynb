```python
import numpy as np
import pandas as pd
import datetime
from CAL.PyCAL import *
###DataAPI.SecTypeRelGet(typeID=u"101001001005",secID=u"",ticker=u"",field=u"",pandas="1")
###DataAPI.MktEqudGet(tradeDate=u"20150513",secID=u"",ticker=u"",beginDate=u"",endDate=u"",isOpen="",field=u"",pandas="1")
###DataAPI.MktStockFactorsOneDayGet(tradeDate=u"20150227",secID=u"",ticker=u"000001,600000",field=u"ticker,tradeDate,pe",pandas="1")
###DataAPI.MktStockFactorsDateRangeGet(secID=u"",ticker=u"000001",beginDate=u"20150101",endDate=u"20150701",field=u"ticker,tradeDate,pe",pandas="1")
###DataAPI.SecTypeRelGet(typeID=u"101001004001001",secID=u"",ticker=u"",field=u"",pandas="1")###nowtime
cal = Calendar('China.SSE')
nowtime = datetime.date.today()
dtt1 = cal.advanceDate(nowtime,'-0B' ,BizDayConvention.Following).strftime('%Y%m%d')
dtt2 = cal.advanceDate(dtt1,'-1B',BizDayConvention.Following).strftime('%Y%m%d')
dtt3 = cal.advanceDate(dtt1,'-2B',BizDayConvention.Following).strftime('%Y%m%d')
```

各种import与时间定义。dtt1是今天的时间，dtt2是前一个交易日。

```python
def qt(x):
    if x>0.75:
        a = 4
    elif x>0.5:
        a = 3
    elif x>0.25:
        a = 2
    else:
        a = 1
    return a
```

把boll带宽度做了四等份。1到4表示现在价位所在boll位置。其中，数字越大，表示处在boll带位置越高。

```python
def zd(x):
    if x>0:
        a = 1
    else:
        a = 0
    return a
```

涨跌分类，0代表跌，代表涨。

```python
def wd(x):
    if x>0.11:
        a = 1
    else:
        a = 0
    return a
```

boll带宽度函数，1表示boll带够宽，关于够宽的意思就是带内至少能容纳两个涨停板。

```python
def rate(ticker = '600749', begin = '20150101',end = dtt2 ):
    a = DataAPI.MktEqudGet(tradeDate=u"",secID=u"",ticker=ticker ,beginDate=begin ,endDate=end ,isOpen="1",field=u"",pandas="1")
    b = DataAPI.MktStockFactorsDateRangeGet(secID=u"",ticker=ticker ,beginDate=begin ,endDate=end ,field=u"ticker,tradeDate,BollDown,BollUp",pandas="1")
    e = [u'secShortName', u'ticker',u'tradeDate', u'BollDown', u'BollUp',u'secID',u'preClosePrice',u'openPrice', u'highestPrice', u'lowestPrice',     u'closePrice',u'accumAdjFactor']
    c = pd.merge(a,b,on = ['ticker','tradeDate'])[e]
    c['Increase'] =100*(c.closePrice - c.preClosePrice)/c.preClosePrice ##涨跌幅
    c['quantile'] = (c.closePrice - c.BollDown)/(c.BollUp - c.BollDown) ##boll带内分位数
    c['wide'] = (c.BollUp-c.BollDown)/(c.BollUp+c.BollDown) ##boll带宽
    c['zd'] = c['Increase'].apply(zd) #涨跌分类
    c['wd'] = c['wide'].apply(wd) #带宽分类
    c['qt'] = c['quantile'].apply(qt)#带分位数分类
    keyname = [u'secShortName', u'ticker', u'tradeDate',u'wide', u'zd', u'qt', u'wd']
    d = c[keyname]
    rate = d['zd'].groupby([d['wd'],d['qt']]).mean()##计算概率
    return rate
    
```

然后就是涨跌概率的计算了。写成了一个函数，默认时间从20150101开始的，默认参数的股票是西藏旅游，别问为什么，随便选的。再验证写一个例子，我就随便找了一只股票。

```python
print rate(),rate('000862')
```
