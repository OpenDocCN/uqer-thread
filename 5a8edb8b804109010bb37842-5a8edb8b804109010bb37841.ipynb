# 春节对沪深300指数的影响

网易新闻：2月22日，狗年首个交易日，A股普涨，三大股指全天飘红，上证综指收涨2.17%，报3268.56点，创一年半最大单日涨幅；创业板指收涨1.88%，报1677.76点；深证成指收涨2.18%，报10658.94点。沪深两市成交3252.75亿元，较上一交易日的2368.59亿元，放量近900亿元。

今天股市迎狗年开门红，A股普涨，沪深300指数收涨2.16%。出于好奇心，把历年春节前后沪深300指数的数据调出以观察规律，发现近10年里，春节后第一个交易日沪深300指数收涨的概率仅为20%，平均收益率为-0.77%，是春节前后10日内最低的。我认为今年的开门红主要原因还是节前股价恐慌性暴跌的反弹。柱状图前5根为节前5个交易日的收益率，后5根为节后5个交易日的收益率。
