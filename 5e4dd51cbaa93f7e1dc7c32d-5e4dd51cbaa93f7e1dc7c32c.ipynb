# 期货升贴水情况202002

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.finance as mpf
ss="20191008"
ee="20200217"
f=['tradeDate','openPrice','closePrice','chgPct']
data=DataAPI.MktMFutdGet(mainCon=u"",contractMark="L0",contractObject=u"IF",tradeDate=u"",\
                         startDate=ss,endDate=ee,field=f,pandas="1")
f=['tradeDate','openPrice','highestPrice','lowestPrice','closePrice','CHGPct']
data2=DataAPI.MktFunddGet(secID=u"510300.XSHG",ticker=u"",tradeDate=u"",beginDate=ss,endDate=ee,field=f,pandas="1")

futuresOpen=data.openPrice
sharesOpen=data2.openPrice*1000

dif=futuresOpen-sharesOpen

futureClose=data.closePrice
sharesClose=data2.closePrice*1000
dif2=futureClose-sharesClose

o=data2.openPrice.astype(float)
h=data2.highestPrice.astype(float)
l=data2.lowestPrice.astype(float)
c=data2.closePrice.astype(float)
source=pd.DataFrame()
day=np.arange(0,len(data2),1)
source['t']=day
source['o']=o
source['h']=h
source['l']=l
source['c']=c
source=source.as_matrix()
fig,ax=plt.subplots(figsize=(10,10))
plt.plot(dif,color='b')
ax2=ax.twinx()
mpf.candlestick_ohlc(ax2,source, colorup='red',colordown='green',width=0.5)
plt.axis('tight')
plt.title('open time dif')
plt.show()

fig,ax=plt.subplots(figsize=(10,10))
plt.plot(dif2,color='b')
ax2=ax.twinx()
mpf.candlestick_ohlc(ax2,source, colorup='red',colordown='green',width=0.5)
plt.axis('tight')
plt.title("close time dif2")
plt.show()
```
