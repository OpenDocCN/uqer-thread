# Pandas中Series的属性和方法汇总

```python
#Series的属性和方法汇总
#Series的属性之.values(读取数据)和.index（读取索引）
from pandas import Series,DataFrame        #导入库
import pandas as pd                         #导入库
print("----------创建一个pandas数据结构----------")
test=Series(["one","two","three"],index=[1,2,3])
print  test
print("-----------开始检验.values属性-----------")
print test.values
print("-----------开始检验.index属性-----------")
print test.index
```

```python
#Series的属性和方法汇总
#Series的方法之.isnull(..)和.notnull(..)判断索引对应的数据是否为空
from pandas import Series,DataFrame        #导入库
import pandas as pd                         #导入库
print("----------创建一个字典----------")
test={1:"one",2:"two",3:"three"}
print  test
print("----------创建带有NaN数值的Pandas----------")
test2=Series(test,index=[1,2,3,4])
print test2
print("----------测试.isnull()方法----------")
print pd.isnull(test2)
print("----------测试.notnull()方法----------")
print pd.notnull(test2)
```
