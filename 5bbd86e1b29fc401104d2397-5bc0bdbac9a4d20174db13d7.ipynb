### 通过风格指数，分析基金风格，排除突发原因（比如大规模申购赎回导致的净值波动）：
### 1. 风格指数（`沪深300成长、沪深300价值、中证500、中证1000、短期债券、中期债券、长期债券、现金`）
### 2. 基金收益率未被风格指数解释的部分认定为`选股alpha`，被风格指数解释的部分认定为`风格alpha`
### 3. `基金超额收益 = 选股Alpha + 风格Alpha`

```python
def fund_case_return(trade_date_list, fundTicker):
    # 获取部分基金收益率
    factor = pd.DataFrame()
    for dt in trade_date_list:
        data = DataAPI.FundNavGet(ticker=fundTicker,dataDate=dt,field=u"ticker,endDate,ADJUST_NAV",pandas="1")
        factor = factor.append(data)
    factor = factor.pivot(index='endDate', columns='ticker', values='ADJUST_NAV')
    factor = factor.pct_change()
    factor.dropna(how='all', axis=0, inplace=True)
    return factor


def fund_daily_return(fundTicker, benchmarkTicker, begin, end):
    # 获取基金和基准的收益率
    daily_ret = pd.DataFrame()   
    data = DataAPI.FundNavGet(ticker=fundTicker,beginDate=begin,endDate=end,
                              field=u"secShortName,endDate,ADJUST_NAV",pandas="1")
    daily_ret['endDate'] = data['endDate'][1:]
    daily_ret[data.iat[0,0]] = data['ADJUST_NAV'].values[1:]*1.0/data['ADJUST_NAV'].values[:len(data)-1]-1 
    
    daily_ret_2 = pd.DataFrame()
    data_2 = DataAPI.MktIdxdGet(ticker=benchmarkTicker,beginDate=begin,endDate=end,
                                field=u"secShortName,tradeDate,closeIndex",pandas="1")
    daily_ret_2['endDate'] = data_2['tradeDate'][1:]
    daily_ret_2[data_2.iat[0,0]] = data_2['closeIndex'].values[1:]*1.0/data_2['closeIndex'].values[:len(data_2)-1]-1
    
    daily_ret = pd.merge(daily_ret, daily_ret_2, how='inner', on='endDate')
    daily_ret.set_index('endDate',inplace=True) 
    return daily_ret

   
def cal_indicators(daily_ret, fre='daily'):
    # 获取基金的评价指标
    if fre == 'daily':
        freq_ = 250
    if fre == 'monthly':
        freq_ = 12
    df_cum_value = (daily_ret + 1).cumprod()
    res = pd.DataFrame(index=['年化收益率','年化标准差','贝塔','夏普比率','信息比率','最大回撤'], 
                       columns=daily_ret.columns, data=0.0)
    
    res.loc['年化收益率'] = (daily_ret.mean() * freq_).apply(lambda x: '%.2f%%' % (x*100))
    res.loc['年化标准差'] = (daily_ret.std() * np.sqrt(freq_)).apply(lambda x: '%.2f%%' % (x*100))
    res.loc['贝塔', daily_ret.icol(1).name] = 1
    res.loc['贝塔', daily_ret.icol(0).name] = np.round((np.cov(daily_ret[daily_ret.icol(0).name],
                                             daily_ret[daily_ret.icol(1).name])[0,1] 
                                             / np.cov(daily_ret[daily_ret.icol(1).name])), 2)
    res.loc['夏普比率'] = (daily_ret.mean() / daily_ret.std() * np.sqrt(freq_)).apply(
                                                                        lambda x: np.round(x, 2))
    res.loc['信息比率', daily_ret.icol(0).name] = np.round(((daily_ret[daily_ret.icol(0).name].mean()
                                                         - daily_ret[daily_ret.icol(1).name].mean())  
                                                         / (daily_ret[daily_ret.icol(0).name]-
                                        daily_ret[daily_ret.icol(1).name]).std() * np.sqrt(freq_)), 2)
    res.loc['最大回撤'] = cal_maxdrawdown(df_cum_value).apply(lambda x: '%.2f%%' % (x*100))
    return res


def cal_maxdrawdown(data): 
    # 获取基金的最大回撤
    if isinstance(data, list):
        data = np.array(data)
    if isinstance(data, pd.Series):
        data = data.values    
    def get_mdd(values): 
        dd = [values[i:].min() / values[i] - 1 for i in range(len(values))]
        return abs(min(dd))
    if not isinstance(data, pd.DataFrame):
        return get_mdd(data)
    else:
        return data.apply(get_mdd)
```

```python
def plot_ret(daily_ret):
    # 画出基金和基准的净值
    daily_ret = daily_ret.dropna()
    sns.set_style("dark")
    daily_ret[daily_ret.icol(1).name] = (1 + daily_ret[daily_ret.icol(1).name]).cumprod() - 1
    daily_ret[daily_ret.icol(0).name] = (1 + daily_ret[daily_ret.icol(0).name]).cumprod() - 1
    
    fig = plt.figure(figsize=(12, 4))
    ax1 = fig.add_subplot(111)
    ax1.grid(axis='y', ls='--')
    ax1.plot(daily_ret.index, daily_ret[daily_ret.icol(1).name], lw=2, color='grey', label='benchmark')
    ax1.plot(daily_ret.index, daily_ret[daily_ret.icol(0).name], lw=2, color='#00BFFF', label='fund')
    ax1.set_title('fund nv')
    ax1.legend(loc=2)

    
def plot_weight(cof_df):
    # 画出基金在各资产风格上的近似比例 
    cof_df.index = pd.to_datetime(cof_df.index)
    fig = plt.figure(figsize=(12, 4))
    ax = fig.add_subplot(111)
    ax.set_ylim([0,1])
    ax.stackplot(cof_df.index, cof_df.values.T, 
                 labels=['BONDS','BONDM','BONDL','HS300G','HS300V','ZZ500','ZZ1000','Cash'], alpha=0.6)
    ax.legend(bbox_to_anchor=(1.16, 1.0),prop=font)
    

def plot_heatmap(df_):
    # 画出相关系数的热力图
    mask_shibor_rate = np.zeros_like(df_.corr(),dtype=np.bool)
    mask_shibor_rate[np.triu_indices_from(mask_shibor_rate)] = True
    camp_shibor_rate = sns.diverging_palette(220,10,as_cmap=True)
    fig = plt.figure(figsize=(15,12))
    sns.heatmap(df_.corr(),mask=mask_shibor_rate,cmap=camp_shibor_rate)
    
    
def plot_strategy(cpt_df):
    # 画出策略和基准的净值
    cpt_df = cpt_df.dropna()
    sns.set_style("white")
    cpt_df['benchmark'] = (1 + cpt_df['benchmark']).cumprod() - 1
    cpt_df['strategy'] = (1 + cpt_df['strategy']).cumprod() - 1
    
    fig = plt.figure(figsize=(10, 4))
    ax1 = fig.add_subplot(111)
    ax1.grid(axis='y', ls='--')
    ax1.plot(cpt_df.index, cpt_df['benchmark'], lw=2, color='grey', label='benchmark')
    ax1.plot(cpt_df.index, cpt_df['strategy'], lw=2, color='#00BFFF', label='strategy')
    ax1.legend(fontsize=16)
    
    
def plot_hedging(cpt_df):
    # 画出对冲后的净值
    cpt_df = cpt_df.dropna()
    sns.set_style("white")
    df_cum_rets = (cpt_df['strategy'] - cpt_df['benchmark'] + 1).cumprod()
    cpt_df['excess_cum'] = df_cum_rets - 1
    running_max = np.maximum.accumulate(df_cum_rets)
    underwater = -((running_max - df_cum_rets) / running_max)
    underwater.index = cpt_df.index

    fig = plt.figure(figsize=(10, 4))
    ax1 = fig.add_subplot(111)
    ax2 = ax1.twinx()
    ax1.grid(axis='y', ls='--')
    ax1.fill_between(underwater.index, 0, np.array(underwater), color='grey', alpha=0.6, label='underwater')
    ax2.plot(cpt_df.index, cpt_df['excess_cum'], lw=2, color='tomato', label='hedging')
    ax2.set_title('hedging & underwater', fontsize=16)
```

```python
def get_target_and_index_ret(start, end, fundTicker):
    # 基金收益率
    daily_ret = pd.DataFrame()   
    data = DataAPI.FundNavGet(ticker=fundTicker,beginDate=start,endDate=end,
                              field=u",endDate,ADJUST_NAV",pandas="1")
    daily_ret['endDate'] = data['endDate'][1:]
    daily_ret['ret'] = data['ADJUST_NAV'].values[1:]*1.0 /data['ADJUST_NAV'].values[:len(data)-1] -1
    fund_ret = daily_ret.set_index('endDate')['ret']
    # 中债1-3年
    zz_13 = DataAPI.MktIdxdGet(tradeDate=u"",indexID=u"",ticker=u"CBA00321",beginDate=start,endDate=end,
                                  field=u"tradeDate,CHGPct",pandas="1").set_index('tradeDate')['CHGPct']
    # 中债5-7年
    zz_57 = DataAPI.MktIdxdGet(tradeDate=u"",indexID=u"",ticker=u"CBA00341",beginDate=start,endDate=end,
                                  field=u"tradeDate,CHGPct",pandas="1").set_index('tradeDate')['CHGPct'] 
    # 中债10年以上
    zz_10 = DataAPI.MktIdxdGet(tradeDate=u"",indexID=u"",ticker=u"CBA00361",beginDate=start,endDate=end,
                                  field=u"tradeDate,CHGPct",pandas="1").set_index('tradeDate')['CHGPct'] 
    # 沪深300成长
    HS300_grouth = DataAPI.MktIdxdGet(tradeDate=u"",indexID=u"",ticker=u"000918",beginDate=start,endDate=end,
                                  field=u"tradeDate,CHGPct",pandas="1").set_index('tradeDate')['CHGPct']
    # 沪深300价值
    HS300_value = DataAPI.MktIdxdGet(tradeDate=u"",indexID=u"",ticker=u"000919",beginDate=start,endDate=end,
                                  field=u"tradeDate,CHGPct",pandas="1").set_index('tradeDate')['CHGPct']
    # 中证500（缺失500成长和价值的数据）                                                                                   
    zz_500 = DataAPI.MktIdxdGet(tradeDate=u"",indexID=u"",ticker=u"000905",beginDate=start,endDate=end,
                                  field=u"tradeDate,CHGPct",pandas="1").set_index('tradeDate')['CHGPct']
    # 中证1000                                                                                   
    zz_1000 = DataAPI.MktIdxdGet(tradeDate=u"",indexID=u"",ticker=u"000852",beginDate=start,endDate=end,
                                  field=u"tradeDate,CHGPct",pandas="1").set_index('tradeDate')['CHGPct']  
    # 现金
    cash = DataAPI.MktRepodGet(secID=u"",ticker=u"131801",beginDate=start,endDate=end,
                                  field=u"tradeDate,closeRate",pandas="1").set_index('tradeDate')['closeRate']
    idx_df = pd.DataFrame()
    idx_df['zz_13'] = zz_13 
    idx_df['zz_57'] = zz_57
    idx_df['zz_10'] = zz_10
    idx_df['HS300_grouth'] = HS300_grouth
    idx_df['HS300_value'] = HS300_value
    idx_df['zz_500'] = zz_500
    idx_df['zz_1000'] = zz_1000
    idx_df['risk_free_rate'] = cash.mean() / 360.0 / 100.0
    idx_df = idx_df.dropna()                                                                             
    return idx_df, fund_ret


def get_optimize_cof(idx_df, fund_ret):
    # 基准指数
    A = np.array(idx_df.values)
    # 基金收益率                                                                              
    Y = np.reshape(np.array(fund_ret.values),(-1,1))
    # 定义目标函数                                                                              
    def func(x):
        bd = np.dot(A, x) - Y.T  
        return np.sum(bd ** 2)
    # 基金在各资产风格上的近似比例之和 不大于1                                                                              
    cons = ({'type': 'eq', 'fun': lambda x:  1 - sum(x)})
    data_len = idx_df.shape[1]
    # 基金在单个资产风格上的近似比例在（0，1）之间                                                                           
    bnds = tuple((0,1) for _ in range(data_len))
    start_pos = np.ones(data_len) *(1.0 / data_len) 
    # 约束求解
    options = {'disp': False, 'maxiter': 500, 'ftol': 1e-20} 
    res = minimize(func, start_pos, bounds=bnds, constraints=cons, method='SLSQP',options=options)
    cof = np.reshape(np.array(res.x),(-1,1))
    # 基金在风格上的收益
    fund_style_ret = A.dot(cof)
    # 基准指数对基金的解释度                                                                              
    r_sq =  1.0 - (np.var(Y-fund_style_ret) / np.var(Y))    
    return res.x, r_sq, fund_style_ret.flatten()
```

```python
def get_fund():
    # 所有基金数据
    fund_sec = DataAPI.SecIDGet(assetClass=u"F",pandas="1")
    fund_info = DataAPI.FundGet(secID=fund_sec.secID,pandas="1")
    # 去除分级基金
    fund_leverage = DataAPI.FundLeverageInfoGet(secID=u"",ticker=u"",exchangeCDLeverage=u"XSHG,XSHE",field=u"",pandas="1")
    fund_info = fund_info[[x not in fund_leverage.secIDLeverage.values for x in fund_info.secID.values]]
    # E-股票型，H-混合型
    fund_info = fund_info[[x in ['H','E'] for x in fund_info.category]]
    # 去除封闭式
    fund_info = fund_info[[x not in ['C'] for x in fund_info.operationMode]]
    # 去除FOF基金 QDII基金 保本基金
    fund_info = fund_info[(fund_info.isFof==0) & (fund_info.isQdii==0) & (fund_info.isGuarFund==0)]
    # 去除海外基金
    fund_info = fund_info[['标普' not in x and '纳斯达克' not in x and '美国' not in x for x in fund_info.secShortName]]
    fund_info = fund_info[['标普' not in str(x) and '纳斯达克' not in str(x) and '美国' not in str(x) for x in fund_info.idxShortName]]
    fund_info = fund_info[['标普' not in str(x) and '纳斯达克' not in str(x) and '美国' not in str(x) for x in fund_info.perfBenchmark]]
    fund_info = fund_info[['MSCI' not in str(x) and '标准普尔' not in str(x) and '全球' not in str(x) for x in fund_info.perfBenchmark]]
    fund_info = fund_info[['欧洲' not in str(x) and '摩根斯坦利' not in str(x) and '富时' not in str(x) for x in fund_info.perfBenchmark]]
    fund_info = fund_info[['德国' not in str(x) and '恒生' not in str(x) and '香港' not in str(x) for x in fund_info.perfBenchmark]]
    fund_info = fund_info[['海外' not in str(x) and '恒生' not in str(x) and '香港' not in str(x) for x in fund_info.perfBenchmark]]
    # 混合型基金
    fund_commingle_info = fund_info[fund_info.category=='H']
    # 普通股票型基金
    fund_equ_info = fund_info[fund_info.category=='E']
    # 只考虑 混合型，普通股票型基金
    fund_all_info = fund_commingle_info.append(fund_equ_info).drop_duplicates('secID')
    return fund_all_info
```

### 3、利用选基指标对所有备选基金进行排序，选取所构造指标最高的5只基金等权构建 FOF组合，按照月频率调仓。

```python
# 以 过去12个月平均选股Alpha 为因子数据
select_alpha_df.index.name = 'endDate'
select_alpha_df.columns.name = 'ticker'
factor_init = select_alpha_df.copy()
del factor_init['002001']
factor_init = pd.rolling_mean(factor_init, window=12)
# 因子需要滞后 1期
factor_init = factor_init.shift(1)
factor_init.dropna(how='all', axis=0, inplace=True)
# 基金月度收益数据
fund_case_ret = fund_case_return(trade_month_list, fundticker_list)
prices_init = fund_case_ret.copy()
del prices_init['002001']
prices_init = prices_init.loc[factor_init.index,:] 
# 数据格式示例
factor_init.ix[[0,-1],:3]
```

```python
def backtest_data(factor_init, fund_num=5, r_square=0.6):
    # 合并因子和收益数据
    pr_un = prices_init.unstack()
    fa_un = factor_init.unstack()
    df_cum = pd.concat([pr_un, fa_un],axis=1)
    df_cum.reset_index(drop=False, inplace=True)
    df_cum = df_cum.rename(columns={0:'ret', 1:'factor'})
    # 按因子值排序
    df_cum = df_cum.sort_values(by='factor', ascending=False)
    df_cum.reset_index(drop=True, inplace=True)
    # 筛选出模型解释度大于0.6的基金作为备选基金
    r_square_un = r_square_df.unstack().reset_index(drop=False)
    r_square_un = r_square_un.rename(columns={'level_0':'ticker','level_1':'endDate', 0:'r_square'})
    r_square_un = pd.merge(df_cum, r_square_un, on=['ticker','endDate'], how='left')
    r_square_un = r_square_un[r_square_un['r_square'] > r_square]
    df_cum = r_square_un[['ticker','endDate','ret','factor']]

    # 每月因子值最大的5只基金的平均收益
    ret = df_cum.groupby(by='endDate')['ret'].apply(lambda x: x.iloc[:fund_num]).to_frame('ret')
    ret.reset_index(drop=False, inplace=True)
    ret = ret.groupby(by='endDate')['ret'].agg('mean').to_frame('strategy')
    # 每月基金的平均收益
    benchmark = df_cum.groupby(by='endDate')['ret'].agg('mean').to_frame('benchmark')

    # 1 以基金均值为基准
    cpt_df = pd.concat([ret, benchmark], axis=1, join='inner')
    cpt_df.index = pd.to_datetime(cpt_df.index)
    # 2 以HS300为基准
    HS300_ret = DataAPI.MktIdxmGet(ticker=u"000300",beginDate=ret.index[0],endDate=ret.index[-1],
                                      field=u"endDate,chgPct",pandas="1").set_index('endDate').sort_index()
    HS300_ret.rename(columns={'chgPct':'benchmark'}, inplace=True)
    cpt_df2 = pd.concat([ret, HS300_ret], axis=1, join='inner')
    cpt_df2.index = pd.to_datetime(cpt_df2.index)
    return cpt_df, cpt_df2, ret
```
