- 先说个人观点吧，在全球经济一体化不断加强的今天，各国彼此间的相互影响越来越大，汇率从原来的固定汇率、到有管理浮动、再到浮动汇率，这个趋势在各个国家都相继进行着，浮动汇率能够满足各国之间资本的自由流动，从全球配置的角度讲确实增加了资源的使用效率，好处是很多的；但在现有情况下，美元霸主地位给各国带来了太大的影响，拿我国和美国对比，我国经济减速，美国经济稳步回升，货币政策出现了背离，这一方面导致我国资本外流、同时也限制了我国的货币政策（难言降息。。。）从历史经验来看，你大哥还是你大哥（美国）

- 既然，现有的状况是全球范围内的大跌，那么我就好奇的`统计了从1991年以来主要国际市场同时出现大跌的时间段`
    - 统计的市场有：中国、美国（大哥）、德国（欧元区代表，二哥）、日本（亚太代表）、香港（兄弟），各国所选取的基准指数在下面有说明
    - 统计时间从1991年开始（大A股刚出世）
    - 统计方法是：统计在过去的20天（不失一般性，一个月）这几个国家的指数都同时出现了超过10%的最大回撤
    - 若在某时间段出现了上述情况，则将日期记录在periods变量中，有三个值，第一个是开始日期，第二个是截止日期，第三个是截止日期后三个月的时间，这样做的目的就是想看若出现这种情况后各国股市后市会怎么走
    - 统计过程如下所示

```python
# 上证综指   000001
# 标普500   SPX
# 德国DAX   GDAXI
# 日经225   N225
# 恒生指数   HSI

import pandas as pd
from datetime import datetime as dt
from matplotlib import pyplot as plt

start_date = '19910101'
tickers = ['000001','SPX','GDAXI','N225','HSI']

data = DataAPI.MktIdxdGet(ticker=tickers, beginDate=start_date, field='ticker,tradeDate,closeIndex').pivot(index='tradeDate',columns='ticker',values='closeIndex')
data = data.fillna(method='pad')

N_days = 20  # 过去N天
drawdown_pct = -0.1  # 回撤比例

periods = []  # 记录回测区间，[start, end, end+N_days]

times = 0  # 控制变量，用来计数
for i in range(N_days,data.shape[0]):
    if times == 0:
        if all(data.iloc[i,:]/data.iloc[i-N_days:i,:].max() - 1 < drawdown_pct):
            if i+3*N_days < data.shape[0]:  # 若是最后一期，则日期没办法向后延续
                periods.append([data.index[i-N_days], data.index[i], data.index[i+3*N_days]])
            else:
                periods.append([data.index[i-N_days], data.index[i], data.index[i]])
            times += 1 
    elif times <= N_days:
        times += 1
    elif times > N_days:
        times = 0
periods[-1][-1] = data.index[-1] #将最后一次的日期延展至最新
```

可以看到，在过去25年的历史中，出现上述情况的次数不多，大部分都在07、08金融危机期间，当然还有15年6月以来的股灾！

#### 废话不多说，先将上面几种情况绘制成图，然后再仔细分析，如下所示：
- 每一个区间段绘制一张图，横坐标是时间，纵坐标是相对区间段第一天的净值，因为各国点数都不一致，可以理解为做了归一化处理

- 每张图都有5条线，分别代表五大指数的走势，时间段是从每一个统计区间的第一天开始，到截止日后的三个月

- 每张图中都有一条垂直的黄线，黄线处对应的时间就是截止日期，也就是说，黄线往左是统计区间，也就是各国都同时发生了大跌，黄线往右是发生大跌后随后三个月的走势

- 前7张图都是历史信息，最后一张图是当前环境

```python
from matplotlib.font_manager import FontProperties
fontP = FontProperties()
fontP.set_size('small')

fig = plt.figure(figsize=(19,12))
for i in range(len(periods)):
    tmp_data = data.loc[periods[i][0]:periods[i][2]]/data.loc[periods[i][0]]
    tmp_ax = fig.add_subplot(3,3,i+1)
    tmp_ax.plot(map(lambda x: dt.strptime(x, '%Y-%m-%d'),tmp_data.index), tmp_data)
    xticks = tmp_ax.get_xticks()
    tmp_ax.set_xticks(xticks[0:-1:2])
    tmp_ax.axvline(x=periods[i][1], linewidth=2.5, color='y')
    tmp_ax.legend(tmp_data.columns, prop=fontP, loc='best')
```
