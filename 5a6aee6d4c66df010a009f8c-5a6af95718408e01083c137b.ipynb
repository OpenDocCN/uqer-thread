**2.我的基本面大局观**
**2.1价值投资的时代来临？**
熟悉量化投资的人肯定会注意到一个现象：以前16年以前很简单的策略可以轻松跑出50%以上的年化收益来，到了16，17年就不行了，甚至亏钱。这里面就有一个基本面的变化在里面，16年以前涨得很好的中小创在17年以后表现不佳。听到很多人在说“A股价值投资”的时代来临了，那么价值投资的时代究竟来临没有？
我们用一个策略来说明，我的策略基本思路是这样的：如果一个股票从低于股票净值上涨到与净值同步，那么我们认为这是一个企业的基本面从坏的情况向好的情况发展的转折点，此时应当积极持有股票。

这个策略里面有几点要说明：这个策略只是为了演示价值投资的思路，因此并没有作太深入优化，卖出策略直接使用盈利减半策略，没有针对大盘回调作风险规避以减少回撤，有兴趣的朋友可以继续优化。这个策略表现平平，但是基本还是跑赢了大盘的，说明我们的思路是很正确的。

**2.2价值投资还是大市值投资**
我们继续深究这个变化，为什么16年以后发生了"价值投资"的这种理念转变呢？有些人说是综合国力的提高，有些人说是监管更加严厉金融市场更加规范。读了一些书，尤其是看了宋鸿兵老是的《宏观》节目之后，经过最近一段时间的思考我觉得最根本的原因是中国机构投资者的资本量发生了根本性的变化，以前主力机构（基金，信托，其他投资机构）的资本量相对较少，所以只能去操纵一些中小盘，而现在主力机构的资本量得到了巨大的提升。尤其是最近两年公募基金经过洗牌之后，资金更加集中在几家大型基金公司的明星基金经理手里，如果去操作小盘股，很容易就引起暴涨暴跌，到时候证监会又下来调查。经过国家队和主力机构这些年的博弈已经形成了一种默契，国家队要维稳，那主力机构就让市场稳定运行，所以你会看到这两年大盘指数前所未有的稳定，权重大盘股涨势喜人，当然又不能涨得过快，于是会有板块轮动。我无法预测这种趋势什么时候结束，我们只能谨慎小心，满怀希望跟着大部队走。
为了验证我的猜想，我把上面的策略稍微修改一下选股的时候加上，LCAP（对数市值），优先考虑对数市值大的，情况是什么样呢？


如果整体来看新策略优势不是很明显， 但是请注意17年1月份1以后的走势你会发现，后面的策略明显优于前者，说明17年以后"大市值"策略的优势更加明显，所以别忘了在你选股的时候加上LCAP因子哦。
