# 基于业绩预告与业绩快报的ST预警

## 摘要
----
最近被ST的股票都好惨呀，之前alpha_cat大神已经写了一篇[《ST股票排雷》](https://uqer.datayes.com/community/share/587c984e23a7d60051a375f2)。其给出的结果已经非常好，本文做一个小小的补充。因为很多公司会公布业绩预告以及业绩快报，所以通过公司的业绩预告与快报，能对公司是否会ST有更多的把握。尤其是业绩快报，基本上已经可以明确公司是否会被ST。并且若直接爬公告的话，公司在发布业绩预告或者业绩快报时，当预测利润为负为导致被ST或暂停上市，在公告中也会提示相应风险的。

## 什么股票会被ST（special treatment）?
---
如果以下条件发生了，会触发股票的ST：
1. 最近两个会计年度的审计结果显示的净利润为负值；

2. 最近一个会计年度的审计结果显示其股东权益低于注册资本;

3. 注册会计师对最近一个会计年度的财产报告出具无法表示意见或否定意见的审计报告；

4. 最近一个会计年度经审计的股东权益扣除注册会计师、有关部门不予确认的部分，低于注册资本；

5. 最近一份经审计的财务报告对上年度利润进行调整，导致连续两个会计年度亏损；

6. 经交易所或中国证监会认定为财务状况异常的

可以看出，2～6条基本无法用来预测股票被ST，但是第一条可以用来预测即将被ST的股票。实际上创业板的股票不会被ST，但是创业板股票上市规则规定，创业板公司若出现连续三年亏损或最近一年净资产为负等11项情形，将暂停上市。换言之，创业板公司在暂停上市之前无须经过*ST的过渡期，但是两年亏损后一样要发布暂停上市风险警示公告的。

## 从业绩预告中发掘
----
对于非ST公司，若公司上一年归属母公司的净利润为负，业绩预告今年也为负，那么显然它被ST的概率很高。

```python
import datetime
import pandas as pd
def st_remove(source_universe, st_date=None):
    """
    给定股票列表,去除其中在某日被标为ST的股票
    Args:
        source_universe (list of str): 需要进行筛选的股票列表
        st_date (datetime): 进行筛选的日期,默认为调用当天
    Returns:
        list: 去掉ST股票之后的股票列表

    Examples:
        >> universe = set_universe('A')
        >> universe_without_st = st_remove(universe)
    """
    st_date = st_date if st_date is not None else datetime.datetime.now().strftime('%Y%m%d')
    df_ST = DataAPI.SecSTGet(secID=source_universe, beginDate=st_date, endDate=st_date, field=['secID'])
    return [s for s in source_universe if s not in list(df_ST['secID'])]
```

首先我们获取`2016-12-30`，所有非ST的A股作为股票池。

```python
universe = set_universe('A')
universe_without_st = st_remove(universe,st_date = '2016-12-30')
```

接着调用业绩预告API获取所有这些股票的2016年预测归属母公司净利润，将预测为负的筛选出来。

```python
eq_mtEf = DataAPI.FdmtEfGet(ticker=u"",secID=universe_without_st,reportType=u"A",endDate=u"2017-04-30",beginDate=u"2016-10-31",publishDateEnd=u"",publishDateBegin=u"",forecastType="",field=u"secID,publishDate,expnIncAPLL,expnIncAPUPL",pandas="1")
eq_mtEf_below_zero = eq_mtEf[eq_mtEf['expnIncAPLL'] < 0]
```

然后调用财务API，获取2015年各股票归属母公司净利润，将为负的筛选出来。

```python
eq_mtIS = DataAPI.FdmtISGet(secID=universe_without_st,reportType=u"A",publishDateEnd=u"20160501",publishDateBegin=u"20160101",field=u"secID,publishDate,endDate,reportType,NIncomeAttrP",pandas="1")  
eq_mtIS = eq_mtIS[eq_mtIS['endDate'] == '2015-12-31']
eq_mtIS_below_zero = eq_mtIS[eq_mtIS['NIncomeAttrP'] < 0]
```

将两个为负的合并到一起得到，有可能被ST的股票。

```python
Ef_IS_summary = pd.merge(eq_mtEf_below_zero,eq_mtIS_below_zero[['secID','endDate','NIncomeAttrP']],on=['secID'])
```

```python
Ef_IS_summary.head()
```

去掉创业板的股票

```python
Ef_IS_summary = Ef_IS_summary[Ef_IS_summary['secID'].apply(lambda x:x[0]).isin(['0','6'])]
Ef_IS_summary.head()
```

现在已经5月3号了，该ST的基本都已经被ST了，我们可以来检验一下，我们预测的准确性。

```python
df_ST = DataAPI.SecSTGet(secID=set_universe('A'), beginDate='20170503', endDate='20170503', field=['secID'])
Ef_IS_summary['是否被ST'] = Ef_IS_summary['secID'].apply(lambda x: 1 if x in df_ST['secID'].tolist() else 0)
Ef_IS_summary
```

```python
print len(Ef_IS_summary[Ef_IS_summary['是否被ST'] == 1]),len(Ef_IS_summary), 1.0 * len(Ef_IS_summary[Ef_IS_summary['是否被ST'] == 1]) / len(Ef_IS_summary)
```

这样筛选出来的65只股票有57只都被ST了，准确度还是非常高的，不过发业绩预告大部分都是1月份的时候了，并且还是有很多公司不发业绩预告的。如果大家想更早的看公司是否会被ST，或者有些公司不发业绩快报，那么可以参考alpha_cat大神的文章，用三季报的利润做一个筛选。

## 从业绩快报中发掘
---
业绩快报相比与业绩预告要正式很多，其不止公布净利润等等，基本上算是一个未经审计的财务报表了。而且它还会直接给出上一年的利润，以及今年未经审计的利润。

我们调用业绩快报API，直接获取今年与去年的归属母公司利润，筛选出两年都亏损的股票，去除创业板的股票。

```python
eq_mtEe = DataAPI.FdmtEeGet(ticker=u"",secID=universe_without_st,reportType=u"A",endDate=u"2017-04-30",beginDate=u"2016-10-31",
                            field=u"secID,publishDate,endDate,NIncomeAttrP,NIncomeAttrPLY",pandas="1")
eq_mtEe_st = eq_mtEe[(eq_mtEe['NIncomeAttrP'] < 0) & (eq_mtEe['NIncomeAttrPLY'] < 0)]
```

```python
eq_mtEe_st_ex3 = eq_mtEe_st[eq_mtEe_st['secID'].apply(lambda x:x[0]).isin(['0','6'])]
eq_mtEe_st_ex3.head()
```
