## 1. 为什么要预测即将被ST股票

根据股社区的统计，股票被ST都是一种利空的现象，主要是因为以下原因：
1. 大量公募基金都有规定不能买入ST股票，保监会规定险资禁止投资ST股。

2. 私募基金可以买ST股票，但是仓位限制更苛刻，普通单支股票你可以买总仓位的20%，单支ST股票你只能买5%仓位

3. ST股票无法质押融资

4. 民间高利贷借钱炒股的合同通常也会有禁止买入ST股票的条款

5. 涨跌幅变小后不利于短线游资炒作

6. 带给公司负面形象

## 2. 什么股票会被ST（special treatment）?
如果以下条件发生了，会触发股票的ST：
1. 最近两个会计年度的审计结果显示的净利润为负值；

2. 最近一个会计年度的审计结果显示其股东权益低于注册资本;

3. 注册会计师对最近一个会计年度的财产报告出具无法表示意见或否定意见的审计报告；

4. 最近一个会计年度经审计的股东权益扣除注册会计师、有关部门不予确认的部分，低于注册资本；

5. 最近一份经审计的财务报告对上年度利润进行调整，导致连续两个会计年度亏损；

6. 经交易所或中国证监会认定为财务状况异常的

可以看出，2～6条基本无法用来预测股票被ST，但是第一条可以用来预测即将被ST的股票。基于本年的3季报的净利润和去年年报的利润，可以大致预测出该股票是否即将被ST。如果今年3季报依然亏损很厉害，那么年报基本上也是亏损的了（要不然利润操纵的也忒明显了～～）。


## 3. 使用优矿的财务数据将这些内容筛选出来

```python
# CQ3代表三季报，A代表年报，下面的API只取三季报和年报
df = DataAPI.FdmtISGet(reportType=u"CQ3,A",publishDateEnd=u"20170101",publishDateBegin=u"20160101",field=u"ticker,secShortName,publishDate,endDate,reportType,NIncome",pandas="1")   

df = df[df['endDate'] >= '2015-12-31']   # 取2015年的年报和2016年的3季报
df = df.sort(['ticker','endDate', 'publishDate','reportType'], ascending=[1,0,0,0])
df = df.drop_duplicates(subset=['ticker', 'endDate','reportType'])  # 因为财报的更正会产生重复的数据，需要去重，只用最新的财报数据
print df.head()
```

```python
import pandas as pd
flat_df = df.groupby('ticker').apply(lambda x:pd.DataFrame(x.values.reshape(1, -1))) # 将数据拍平
flat_df = flat_df.iloc[:, [0,1,5,11]]  # 选取需要的列
print flat_df.head()
```

得到原始的数据后，首先要把2016年3季报和2015年年报的净利润都是负值的股票筛选出来，ST大概率会在这些股票中产生！

```python
flat_df = flat_df.reset_index(drop=True)
flat_df.columns = ['ticker', 'secName', 'cq3','a'] # cq3代表3季报，a代表年报
income_df = flat_df[(flat_df['cq3']<0) & (flat_df['a']<0)]  # 2016年3季报和2015年年报的净利润都是负值
print income_df.head()
```

通过上面可以看到有些股票已经被ST了，需要被滤除：

```python
# 滤除已经被ST的股票
from datetime import date
today = date.today().strftime("%Y%m%d")
# 下面的方法虽然可以滤除现在的ST股票，但是还会残留退市的股票
# st_df = DataAPI.SecSTGet(secID=set_universe("A", today), beginDate=today, endDate=today, field=['secID'])
# income_df = income_df[~income_df['ticker'].isin(st_df['secID'].tolist())]
income_df = income_df[~income_df['secName'].str.contains("ST", na=False)]
```

还有些股票是B股的，这里暂时不予考虑，也滤除B股的股票，只留下我大A股

```python
# 滤除B股并且排序
income_df = income_df[~(income_df['ticker']> '900000')]
income_df = income_df[~((income_df['ticker']>='200000') & (income_df['ticker']<'300000'))]
income_df = income_df.sort(['cq3','a'], ascending=[0,0])
income_df.columns = [u'ticker',u'股票名称',u'2016三季报利润',u'2015年报利润']
income_df = income_df.reset_index(drop=True)
print income_df.head()
```

加上2016年报发布的时间，也就是有可能地雷引爆的那一天！

```python
#加上年报的预披露时间
import pandas as pd
report_time = pd.read_csv('annual_report/report_time.csv',converters={'ticker_symbol': str},sep='\t')
report_time.columns = ['ticker','2016年年报发布时间']
result = income_df.merge(report_time, on='ticker')
print result.to_html()
```

上面表格中的排序是按照3季度的净利润大小排序的，亏损的越多，排名越靠近底部，那么被ST的概率越大。像重庆钢铁这种，基本上跑不掉了。排名靠前的概率就比较低了，因为3季报亏损的较少，4季度有可能就挣钱填上了，实在不济，在年报中将利润稍作操纵，即可摆脱被ST的命运。
