```python
a=DataAPI.MktIdxdGet(indexID=u"",ticker=u"000001",tradeDate=u"",beginDate=u"20161231",endDate=u"20171231",exchangeCD=u"XSHE,XSHG",field=u"ticker,tradeDate,closeIndex,CHGPct",pandas="1")
r=a.CHGPct.cumsum().iloc[-1]  #注释：iloc[-1]取得是最后一行数据
print '年收益率：'
print r
print ''
returns=a.pivot(index='tradeDate',columns='ticker',values='CHGPct')
returns = returns.fillna(value=0)
cov1 = returns.cov() * 252
print '年标准差：'
cov=cov1.iloc[-1]
print cov
print ''

df1=DataAPI.MktIdxFactorDateRangeGet(secID=u"",ticker=u"000001",beginDate=u"20171229",endDate=u"20180604",field=u"ticker,tradeDate,PE1",pandas="1")
print '上证综指市盈分布：'
print df1.iloc[-1]
print df1.describe()
print ''
```
