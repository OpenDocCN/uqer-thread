```python
#选股
import numpy as np
import pandas as pd
from pandas import *
from statsmodels.tsa.stattools import adfuller 
from statsmodels.tsa.stattools import *
from statsmodels.tsa.stattools import coint
import statsmodels.api as sm
import statsmodels.tsa.stattools as ts

beginDate="20171202"
endDate="20180502"
universe = set_universe(IndSW.FangDiChanKaiFaL3, '20160502')
print universe

def DfTest(data):
    adftest =adfuller(data)
    result = pd.Series(adftest[0:4], index=['Test Statistic','p-value','Lags Used','nobs'])
    for key,value in adftest[4].items():
        result['Critical Value (%s)'%key] = value
    return result

sec = []
#板块内循环，原始数据肯定不平稳了，直接一阶差分，通过P值过滤掉非平稳
for i in range(len(universe)):
    data1=DataAPI.MktEqudAdjGet(tradeDate=u"",secID=universe[i],ticker=u"",beginDate=beginDate,endDate=endDate,isOpen="",field=u"closePrice",pandas="1")
    diffx=data1.diff(1)
    diffx.dropna(inplace=True)
    diffx=np.array(diffx).T[0]
    zz = ts.adfuller(diffx)
    if zz[1] < 0.05:
        sec.append(universe[i])
    
print sec
#验证平稳后的时间数据序列两两验证协整关系
for i in range(5):
    for j in range(i+1,5):
        data1=DataAPI.MktEqudAdjGet(tradeDate=u"",secID=universe[i],ticker=u"",beginDate=beginDate,endDate=endDate,isOpen="",field=u"closePrice",pandas="1")
        data2=DataAPI.MktEqudAdjGet(tradeDate=u"",secID=universe[j],ticker=u"",beginDate=beginDate,endDate=endDate,isOpen="",field=u"closePrice",pandas="1")
        x=np.array(data1)
        y=np.array(data2)
        x=x.T[0];y=y.T[0]
        a,pvalue,b = coint(x,y)
        if pvalue < 0.05:
            print universe[i],universe[j],pvalue
```
