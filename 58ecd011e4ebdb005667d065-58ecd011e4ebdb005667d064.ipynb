## fama的多因素模型
fama在 A five-factor asset pricing model论文里面提了5种因素
- 市场组合收益率
- 公司市值（即SML，small minus large）
- 账面市值比（即HML，high minus low）
- 盈利能力（profitability）因子
- 投资模式（investment patterns）因子（即CMA，conservative minus aggressive）

分类操作方法：
- 分别根据市值-账面市值比、市值-营业利润率和市值-投资水平，对股票进行3次2×3划分，每次得到6个投资组合。以市值-账面市值比划分为例，将市值以纽交所均值为分水岭，划分为大、小2类；对账面市值比，则以纽交所的第30和第70百分位数为分水岭，划分为高、中、低3类
- 分别根据市值-账面市值比、市值-营业利润率和市值-投资水平，以纽交所均值为分水岭，对股票进行3次2×2划分，每次得到4个投资组合。
- 根据市值-账面市值比-盈营业利润率-投资水平，对股票进行1次2×2×2×2的划分，得到16个投资组合.       


fama比较认同第三种方法,根据市值-账面市值比-盈营业利润率-投资水平，对股票进行1次2×2×2×2的划分，得到16个投资组合,有效地从平均收益率中，分离出市值、账面市值比、营业利润率和投资水平的风险溢价
<br \>
假设分组内的股票是等权重买入卖出的，则分组股票组合的收益率可以通过分组内股票收益率的算术平均值获得。
根据因素排序形成的分组股票组合的收益率差别设定为该因素的收益率，fama因素模型将组合收益率分解成多个因素收益率的线性分解函数。

对比前面的CAPM模型的一元线性回归，Fama 3因子模型和5因子模型采用APT多因素模型的多元线性回归。
就是将收益率$r_i$拆分成多个因素收益率的线性函数
$R_i(t)–R_F(t) = a_i + b_i(R_M(t) – R_F(t))+ s_iSMB_t + h_iHML_t + e_i$
$R_i(t) – R_F(t) = a_i + b_i(R_M(t) – R_F(t)) + s_iSMB_t + h_iHML_t + r_iRMW_t + c_iCMA_t + e_i$
通过使用线性回归确定不同股票组合对因素收益率的分解系数解释收益率，也就是定价模型。

由于A股和美股的差别就像中国和美国国情的差别那么大，所以对于fama得到的结论这里不做探讨。

Fama的多因子模型是经典的多因素定价运用范例，论坛里面已经有很多佳作，这里不重复灌水了。
