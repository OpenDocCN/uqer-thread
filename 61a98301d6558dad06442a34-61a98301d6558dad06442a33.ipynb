我一般粗略的把交易行为分为基本面炒股法和技术面炒股法，
基本面炒股就是利用各个财务因子、市值因子等反应企业盈利情况的数据做依据，进行选股和交易。
技术面炒股就是利用形态、技术指标位依据进行选股和交易。

这两种方式，更容易实现的是基本面，比如净利润增长率这个因子，正常情况下，数值越大，股价更容易上涨，这种关系是更容易直观反映到股价上的。
而技术面的情况要复杂的多，如最简单的均线金叉，往往还要借助非常多的其他指标辅助判断。由于无法穷举所有的情况，因此这一类规则较难通过量化的方式实现（也许是我技术太渣）。

重点说下基本面的方法制作机器学习模型。
