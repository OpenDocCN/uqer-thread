# 二.市场强弱状态
- ### 我们根据当前所有申万一级行业指数中站上20日均线的行业的数目来划分市场强弱状态，因为指数站上20日均线通常是中短期强势走势的开端
- ### 如果有70%以上的行业指数在20日均线以上，则当前为强市
- ### 如果有30%以下的行业指数在20日均线以上，则当前为弱市
- ### 如果有30%-70%的行业指数在20日均线以上，则当前为震荡市
- ### 下面就是编程实现：

```python
#加载需要使用的包
from __future__ import division
import numpy as np
import pandas as pd
import scipy.stats as st

from CAL.PyCAL import *    # CAL.PyCAL中包含font
font.set_size(22)
small_font = font.copy()
small_font.set_size(12)

import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('white')

import math
import datetime

import talib
```

```python
class MktSI(object):
    
    def __init__(self,N,begindate,enddate,universe_name='000300.ZICN'):
        ###----传入参数----
        self.N = N                  #取前N日的数据作为依据
        self.begindate = begindate  #开始时间
        self.enddate = enddate      #结束时间
        self.universe_name = universe_name  #股票池
        ###----自定义参数----
        
        self.datelist = self.get_datelist()  #传入开始和结束日期，得到交易日列表list
        self.industrydata = self.get_industrydata()    #获取行业指数信息
        self.MktSIdata,self.strength,self.weak,self.wave = self.get_MktSI()
    
    ###得到日期列表
    def get_datelist(self):   
        #从dataapi调取交易日序列
        datedata=DataAPI.TradeCalGet(exchangeCD=u"XSHG",beginDate=self.begindate,endDate=self.enddate,field=['calendarDate','isOpen'])
        datedata=datedata[datedata['isOpen'] == 1]
        date= map(lambda x: x[0:4]+x[5:7]+x[8:10], datedata['calendarDate'].values.tolist())
        return date
    
    ###得到日期内的行业指数数据
    def get_industrydata(self):
        industrydata = pd.DataFrame(columns=['tmp'],index=self.datelist,data=0)
        industryinf = DataAPI.IndustryGet(industryVersionCD=u"010303",industryLevel=u"1",isNew=u"1",field=u"indexSymbol").indexSymbol.tolist()
        industryinf = [str(x) for x in industryinf]
        for x in industryinf:
            data=DataAPI.MktIdxdGet(ticker=x,beginDate=self.begindate,endDate=self.enddate,field=u"tradeDate,closeIndex").set_index('tradeDate')
            data.index = map(lambda x: x[0:4]+x[5:7]+x[8:10], data.index.tolist())
            data.columns = [x]
            industrydata = pd.concat([industrydata,data],axis=1)
        #industrydata = industrydata.dropna()
        del industrydata['tmp']
        return industrydata
    
    ### 查看某一天是强市还是弱市
    def get_MktSI_daily(self,date):
        count = 0           #超过均线的行业数
        industry_count = 0  #总行业数
        industrydata = self.industrydata
        industrydata.index = np.arange(len(industrydata))
        if date not in self.datelist:
            #print 'wrong date'
            return np.nan
        else:
            index_1 = self.datelist.index(date)
            if index_1-self.N+1 < 0:
                #print 'wrong date'
                return np.nan
            else:
                index_2 = index_1-self.N+1
                for x in industrydata.columns:
                    tmp = industrydata.ix[index_2:index_1+1,[x]]
                    tmp = tmp.dropna()
                    if len(tmp) >0 :
                        industry_count +=1
                        if tmp.iat[len(tmp)-1,0] >= tmp[x].mean():
                            count += 1
        return count/industry_count
    
    ### 得到指定日期内的市场强弱信息
    def get_MktSI(self):
        MktSIdata = pd.DataFrame(data=self.datelist,columns=['tradedate'])
        MktSIdata['MktSI'] = map(lambda x:self.get_MktSI_daily(x),self.datelist)
        MktSIdata = MktSIdata.dropna()
        strength = MktSIdata[MktSIdata['MktSI']>0.7]['tradedate'].tolist()
        weak = MktSIdata[MktSIdata['MktSI']<=0.3]['tradedate'].tolist()
        wave = MktSIdata[(MktSIdata['MktSI']>0.3) & (MktSIdata['MktSI']<=0.7)]['tradedate'].tolist()
        return MktSIdata,strength,weak,wave
    
    ### 画个图显示一下，市场强弱划分的情况
    def get_plot(self):
        index_data = DataAPI.MktIdxdGet(indexID=self.universe_name,beginDate=self.begindate,endDate=self.enddate,field=u"tradeDate,closeIndex")
        #index_data.set_index('tradeDate',inplace=True)
        index_data['tradeDate'] = map(lambda x: x[0:4]+x[5:7]+x[8:10], index_data['tradeDate'].values.tolist())
        index_data.columns = ['tradeDate',self.universe_name]
        
        fig = plt.figure(figsize = (16,6))
        ax = fig.add_subplot(111)
        index_data[self.universe_name].plot(ax=ax,title=u'市场强弱划分',color='y')
        if len(self.strength)>0:
            for t in self.strength:
                tmp = index_data[index_data['tradeDate'] == t]
                plt.scatter(tmp.index.tolist(),tmp[self.universe_name].values.tolist(),color='r',linewidths=1.5)
                #index_data.ix[self.strength,:].plot(ax=ax,title=u'市场强弱',color='r')
        if len(self.weak)>0:
            for t in self.weak:
                tmp = index_data[index_data['tradeDate'] == t]
                plt.scatter(tmp.index.tolist(),tmp[self.universe_name].values.tolist(),color='g',linewidths=1.5)
                #index_data.ix[self.weak,:].plot(ax=ax,title=u'市场强弱',color='g')
        if len(self.wave)>0:
            for t in self.wave:
                tmp = index_data[index_data['tradeDate'] == t]
                plt.scatter(tmp.index.tolist(),tmp[self.universe_name].values.tolist(),color='blue',linewidths=1.5)
                #index_data.ix[self.wave,:].plot(ax=ax,title=u'市场强弱',color='b')
        #ax.legend(map(lambda x: x.decode('utf-8'),data.columns.values),prop=small_font,loc='best')
        ax.set_xlabel(u'交易日期', fontproperties=small_font)
        M = int(len(index_data)/10)
        plt.xticks([index_data.index[j*M] for j in range(int(len(index_data)/M))]+[index_data.index[-1]],\
                   [index_data.tradeDate[k*M] for k in range(int(len(index_data)/M))]+[index_data.tradeDate[len(index_data)-1]],\
                   fontproperties = small_font, rotation = 70)
        ax.title.set_font_properties(font)
        ax.set_axis_bgcolor('white')
        ax.grid(color='gray', alpha=0.3, axis='x')
```

# 三.构建策略
- ### 下面依据此来构建一下策略，正所谓因子择时，其实这里择的是市场的状态。我们不妨先考察最简单的动量反转因子的效果
- ### 策略逻辑：每个调仓日，观察上一交易日的市场状态：若上一交易日为强市，则等权买入前一交易日比前T交易日的收益为正的股票；若上一交易日为弱市，则卖空所有股票；若上一交易日为震荡市，则不操作
