在Python里，保留字不能用作常数或变数，或任何其他标识符名称。所有 Python 的关键字只包含小写字母。

####保留字-and
在Python中，and执行逻辑运算。但是不同于C/Java，Python中and运算的结果不是布尔值，而是参与运算的值之一。
and在运算中是执行自左到右的逻辑运算，如果运算过程中的所有值都是真，则返回最后一个真值；否则返回逻辑运算过程中的第一个假值。

```python
v0 = 1 and 2
print v0
v1 =  0 and 2
print v1
v2=1 and 2 and 3  
print v2
```

####保留字-or
在Python中，or执行逻辑运算。同and运算符一样，or运算的结果同样是参与运算的值之一。
or在运算中是执行自左到右的逻辑运算，如果运算过程中的所有值都是假，则返回最后一个假值；否则返回逻辑运算过程中的第一个真值。

```python
v0 = 1 or 2
print v0
v1 =  0 or 2
print v1
v2=1 or 2 or 3  
print v2
```

####保留字-break
在Python里,break语句用来终止循环语句，即循环条件没有False条件或者序列还没被完全递归完，也会停止执行循环语句。如果使用嵌套循环，break语句将停止执行最深层的循环，并开始执行下一行代码。其中，break语句常用在while和for循环中。

```python
for i in range(10):
    if i == 4:
        break
    print "current number:",i
  
       
```

####保留字-continue
在Python里,continue 语句跳出本次循环，而break跳出整个循环。continue 语句用来告诉Python跳过当前循环的剩余语句，然后继续进行下一轮循环。

```python
for i in range(10):
    if i == 4:
        continue
    print i,
```

###2.3 列表
 在Python里，列表是使用最广泛的数据结构。列表中的每个元素都分配一个数字 - 它的位置，或索引，第一个索引是0，第二个索引是1，依此类推。列表是写在方括号里面，用逗号隔开的元素列表。
列表基本操作：
    1. 列表创建
    2. 列表访问
    3. 列表更新
    4. 列表删除
常见的列表操作方法：
    a.列尾添加
    list.append(obj)
    b.列表扩充
    list.extend(seq)
    c.查询某个元素在列表中出现的次数，obj任意类型
    list.count(obj)
    d.从列表中找到某个值第一次出现位置的索引
    list.index(obj)
    e.在指定索引位置，添加一个元素
    list.insert(index,obj)
    f.移除列尾的元素,可以指定index索引位置，并返回移除的元素
    index默认为列尾的索引下表
    list.pop(index)   等同于del list[index]
    g.移除指定某个值第一次匹配项
    list.remove()
    h.反向列表
    list.reverse()
    i.排序
    list.sort()  #快排
    #冒泡排序、选择排序、插入排序、快速排序、(堆排序)、归并排序
    函数
    (1)max(list)、min(list)
    (2)list(seq) 序列转换列表化（unicode字符串无法直接列表化）
    （3）cmp(list1,list2) 大于返回1，等于返回0，小于返回-1


```python
#列表创建
list1 = [12,21,23,21,34,45,56,78,"hello","python"]
list2 = [12,'he',"we",89]
list3 =["h","e","l","l"]
print list1
print list2
print list3

# 列表访问
print list1[:]
print list1[0]
print list1[-1]
print list1[0:5]
# 列表追加
print list1.extend(list2)
# 获取列表长度
print "列表1的长度是",len(list1)
# 删除列表元素
del list1[0]
print list1
# 元素是否存在于列表中
print 'we' in list2
# 迭代
for i in list1:
    print i
# 比较两个列表的元素
print cmp(list1, list2)
# 列表最大/最小值
print max(list1)
print min(list2)
# 在列表末尾添加新的元素
list2.append("she")
print list2
# 统计某个元素在列表中出现的次数
print list1.count(21)
# 从列表中找出某个值第一个匹配项的索引位置
print list2.index('we')
# 将对象插入列表
list2.insert(0, 'you')
print list2
# 移除列表中的一个元素（默认最后一个元素），并且返回该元素的值
print list3.pop(0)
print list3
# 移除列表中某个值的第一个匹配项
list3.remove("l")
print list3
# 反向列表中元素
list3.reverse()
print list3
# 对原列表进行排序
list3.sort()
print list3
```

###2.5 字典
 在Python里，字典是另一种可变容器模型，且可存储任意类型对象。字典的每个键值(key=>value)对用冒号(:)分割，每个对之间用逗号(,)分割，整个字典包括在花括号({})中。字典格式：d = {key1 : value1, key2 : value2 }。
字典基本操作：
    1. 字典创建
    2. 字典访问
    3. 字典更新
    4. 字典删除  a.清空字典，保留字典的空间 dict.clear() b.删除字典,不保留字典空间 del dict
字典特性：
    a.不允许同一个键出现多次，以最后一个值为准
    b.键不可变，所以可以为数字、字符串、元祖，但是不能为列表
常见的字典操作方法：
    a.浅复制
    dict.copy() 
    b.创建一个新的字典，以序列seq中元素作为字典键，val为所有键的默认值
    dict.fromkeys(seq[,val])
    c.返回指定的键的值，如果键不存在，返回default值
    dict.get(key,default)
    d.判断某个键是否存在
    dict.has_key(key)
    e.返回所有的键名，以列表形式
    dict.keys()
    f.以列表形式返回可遍历的（键值）元祖数组
    dict.items()
    g.以列表形式返回所有的值
    dict.values()
    h.get()升华版，假如键不存在，则增加键设置值为default
    dict.setdefault(key,default)
    i.字典2追加到字典1中，并且执行更新操作
    dict1.update(dict2)

```python
d1 = {"a":12,"b":23,"c":34}
d2 = {1:"Mary", 2:"Mike", 3:"Jack"}
#浅复制
d3 = d1.copy()
print d3
#创建一个新的字典
d4 = d1.fromkeys([1,2,3,4],"value")
print d4
#字典访问
d1.get("b")
#判断某个键是否存在
d1.has_key(23)
#返回所有的键名，以列表形式
print d1.keys(),d2.keys()
#以列表形式返回可遍历的（键值）元祖数组
print d1.items(),d1.values()
#字典追加
d1.update(d2)
print d1
```

###三、控制语句与函数

###1.控制语句
在Python里，控制语句一般包括条件语句和循环语句。Python条件语句是通过一条或多条语句的执行结果（True或者False）来决定执行的代码块。Python循环语句是多次执行一条或多条语句的执行语句。
###1.1 条件语句
Python程序语言指定任何非0和非空（null）值为true，0或者null为false。Python 编程中if语句用于控制程序的执行，基本形式为：
                            if 判断条件：
                                执行语句……
                            else：
                                执行语句……

```python
#两段式
if True:
    print "这是真的。"
else:
    print "不真。"
#多段式
 
a = 12    
if a > 12:          
    print "比12大z"       
elif a == 12:
    print "等于12."
else:
    print "比12小。"     
```

###1.2 循环语句
Python循环语句一般包括for循环和while循环。

```python
#while循环
i = 0
while i <= 10:
    print i,
    i += 1
#for循环
print "\n"
for i in range(11):
    print i,
```
