# 风险度量

通过计算每一天的风险值VAR（参考‘期权期货及其他衍生品’）来衡量市场的风险
这里收益率波动率利用GRACH模型进行预测

```python
#股票日波动率为vol
#前一天收盘价为preClosePrice
#有precent把握肯定，未来futureDay内价格下跌不会超过N^-1(1-percent)倍的标准差
from scipy.stats import norm
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter, MaxNLocator 
import math
import arch
import matplotlib.pyplot as plt
from scipy import  stats
import statsmodels.api as sm
import numpy as np
from arch import arch_model

def SingleAssertVAR(vol,preClosePrice,percent,futureDay):
    return vol*preClosePrice*norm.ppf(1-percent)*math.pow(futureDay,0.5)
def CompoundAssetsVAR(volList, covarianceMatrix):
    volArray = np.mat(volList)
    return (volArray*covarianceMatrix*volArray.T)[0,0]
#print(CompoundAssetsVAR([1,2,3],np.array([[1,2,6],[2,3,6],[3,4,6]])))

#通过grach模型处理从endDay-deltaDay到endDay这deltaDay个交易日的收益率，来预测之后一天的股票stock波动率
def volPredict(endDate,deltaDay,stock='',IsIndex = False):
    IndexDate = list(DataAPI.MktIdxdGet(ticker=u"000001",beginDate=u"20070101",endDate=endDate,field=u"tradeDate")['tradeDate'])
    beginDate = IndexDate[len(IndexDate)-deltaDay]
    if(IsIndex):
        StockData = DataAPI.MktIdxdGet(ticker=u"000001",beginDate=beginDate,endDate=endDate,field=u"tradeDate,closeIndex").set_index('tradeDate')
        StockData['returns'] = (StockData['closeIndex']-StockData['closeIndex'].shift(1))/StockData['closeIndex'].shift(1)
    else:
        
        StockData = DataAPI.MktEqudGet(ticker=stock,beginDate=beginDate,endDate=endDate,field=u"tradeDate,closePrice").set_index('tradeDate')
    
        StockData['returns'] = (StockData['closePrice']-StockData['closePrice'].shift(1))/StockData['closePrice'].shift(1)
    StockData = StockData.dropna()
    returns = list(StockData['returns'])
    t = sm.tsa.stattools.adfuller(returns)  # ADF检验
    print "p-value:   ",t[1]
    
    
    
    #于是我们对均值建立AR(1)模型
    order = (1,0)
    model = sm.tsa.ARMA(returns,order).fit()
    
    #考虑波动率
    at = returns -  model.fittedvalues
    at2 = np.square(at)
    
    
    #于是我们对波动率建立AR(1)模型
    
    #根据我们之前的分析，可以粗略选择**均值模型为AR(1)模型**，**波动率模型选择GARCH(1,1)模型**
    am = arch.arch_model(returns,mean='AR',lags=1,vol='GARCH') 
    res = am.fit()
    omega = res.params[2]
    alpha = res.params[3]
    beta = res.params[4]
    ini = res.resid[-1:]
    a = np.array(res.params[1:2])
    w = a[::-1] # 系数
    new_error = returns[-1] - (res.params[0] + w.dot(ini[-1:]))
    at_pre = new_error**2
    ini2 = res.conditional_volatility[-1] #上一个条件异方差值

    newPre = omega + alpha*at_pre + beta*ini2
    return newPre

def drawEveryDayIndexVar():
    IndexDate = list(DataAPI.MktIdxdGet(ticker=u"000001",beginDate=u"20150101",endDate='20170601',field=u"tradeDate")['tradeDate'])
    VARList = []
    closePrice = []
    dayList = []
    for day in range(len(IndexDate)-1):
        tradeDate = IndexDate[day+1]
        beforeDate = IndexDate[day]
        preClosePrice = DataAPI.MktIdxdGet(ticker=u"000001",tradeDate=beforeDate,field=u"closeIndex")['closeIndex'][0]
        vol = volPredict(beforeDate,60,IsIndex = True)
        VAR = SingleAssertVAR(vol,preClosePrice,0.95,5)
        VARList.append(VAR)
        closePrice.append(DataAPI.MktIdxdGet(ticker=u"000001",tradeDate=tradeDate,field=u"closeIndex")['closeIndex'][0])
        dayList.append(tradeDate)
    newDay = []
    for day in dayList:
        newDay.append(day)
    x= range(len(newDay))
    def format_fn(tick_val, tick_pos):
        if int(tick_val) in x:
            return newDay[int(tick_val)]
        else:
            return 
    fig = plt.figure(figsize=(24,6))
    ax1 = fig.add_subplot(111)
    ax2 = ax1.twinx()
    ax1.xaxis.set_major_formatter(FuncFormatter(format_fn))
    ax1.xaxis.set_major_locator(MaxNLocator(integer=True))
    ax2.xaxis.set_major_formatter(FuncFormatter(format_fn))
    ax2.xaxis.set_major_locator(MaxNLocator(integer=True))
    ax1.plot(range(len(dayList)),closePrice,'r')
    ax2.plot(range(len(dayList)),VARList,'b')
    plt.show()
drawEveryDayIndexVar()
```

采用VAR的方式可以很好的刻画出2015年的风险，并且2017年初的风险与2015年的风险有两个值很相近，说明近期投资应当谨慎
