> **QuartzFutures 特点**

优矿期货策略回测框架(QuartzFutures)有`以下特点`：

1、QuartzFutures 是Python 2.7 环境下的一个标准化的量化投资策略回测框架，`支持日线、分钟线和秒线的期货量化投资策略`。依照QuartzFutures 的策略模板，您可以开发和检验各种独特的期货投资策略，并方便地进行回测，了解投资策略在市场中的历史表现，比较不同投资策略的优劣。

2、QuartzFutures `兼容`各类Python的数学、统计、金融和机器学习等扩展模块，如Pandas, Numpy, Scipy, statsmodels, Scikit Learn, talib等等，具体可以参考优矿官网【常见问题】中的优矿支持的Python模块章节。

3、QuartzFutures `覆盖了策略研究的所有阶段`。其集成了数据获取功能，使用的是通联数据的数据API，能够在线迅速高效地获取证券行情数据、指数行情数据和无风险收益率数据等，快速进行策略的回测。
 
4、QuartzFutures 运行在`高度安全`的优矿平台上，具体可以查看优矿官网【常见问题】中的安全章节。
