## 策略的理论基础

  > 一旦市场形成了下降（或上升）的趋势后，就将沿着下降（或上升）的方向运行。

  ### 之前做过MA5，MA10的金叉，效果并不是十分理想，主要的原因是假金叉，均线的粘合等问题。

  ### 因此我用了更大周期的均线，采用MA10，MA20来判别大趋势。

  ### 另外如果前9天出现MA20>MA10，则不认为这个金叉有效（粗略的认为这是粘合了）

![nianhe](http://www.xftz.cn/uploads/allimg/150422/6938-1504221323395A.png)

  ### 类似黄色框框中的就是粘合，如果前9日都是MA10< MA20，那么这时候的MA10>MA20我才认为是有效金叉。

## 买入条件
***

  - 1、有效金叉（上文提到了）

  - 2、上一交易日MA20/上上交易日MA20的比例大于99.5%。（原本是大于100%，也就是说MA20必须方向向上，但由于MA20反应较为滞后，我调节成了99.5%。）

  - 3、上一交易日的收盘价小于最近10个交易日最高点90%。(下跌过的金叉才值钱，如果之前的走势很平缓，那么金叉的意义就不大了)

## 卖出条件
***

  - 1、止损5%

  - 2、止盈20%

  - 3、10日内最大价格-最小价格/平均价格 小于2%。（最近10天波动小于2%，那么我认为这个股就有点占着那什么不拉那什么了。所以把他清出去）

## 交易目标 全A股
***

  ### 因为满足条件的股票非常少，所以把目光放到全A股上

## 仓位的控制
***

  ### 因为满足条件的股票比较少，为了防止单支股票带来影响太大。我每次取10% 和 95%/买入列表中的最小值来作为仓位的依据。这样如果只有2支股票满足条件，每只只买入10%的仓位。如果有20只，那么用95%/20得出买入的比例


  ### 这个策略不是ALPHA对冲策略，不需要对冲股指。
  ### 我认为卖出条件还有优化的空间
