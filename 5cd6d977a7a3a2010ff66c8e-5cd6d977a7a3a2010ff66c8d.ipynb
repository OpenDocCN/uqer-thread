# 集成学习拟合因子和股价

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import datetime
import seaborn as sns
import sklearn.ensemble 
pd.set_option('display.max_rows', 100)

#获取000001因子
Factors=DataAPI.MktStockFactorsDateRangeGet(secID=u"",ticker=u"000001",beginDate=u"20120101",endDate=u"20190429",field=u"",pandas="1")

#获取00001历史价格
Price=DataAPI.MktEqudGet(secID=u"",ticker=u"000001",beginDate=u"20120101",endDate=u"20190429",isOpen="",field=u"",pandas="1")

#去除与Price重复的列，以便合并
Factors=Factors.drop(labels=["secID","ticker"],axis=1)
Price=Price.drop(labels=["secID","ticker","secShortName", "exchangeCD"],axis=1)

#去除因子中全为空值的列
for i in Factors.columns:
    if Factors[i].isnull().sum()==Factors.shape[0]:
        Factors=Factors.drop(i,axis=1)

#合并因子和股价，以便进行集成学习及相关分析
Factors=Factors.merge(Price[['closePrice','tradeDate']],on='tradeDate',how='left')

#转换tradeDate为日期类型，以便进行日期索引，同时删除index列
Factors['tradeDate']=pd.to_datetime(Factors['tradeDate'],format="%Y-%m-%d")
Factors.index=Factors['tradeDate']
Factors=Factors.drop(["tradeDate"],axis=1)

#去除空值
Factors=Factors.dropna()

print Factors.columns
print Factors.head()
```

```python
#集成学习函数列表
function_list=[sklearn.ensemble.AdaBoostRegressor(),sklearn.ensemble.BaggingRegressor(),sklearn.ensemble.ExtraTreesRegressor(),sklearn.ensemble.GradientBoostingRegressor(),sklearn.ensemble.RandomForestRegressor()]

#集成学习函数名称，作图用
name_list=["AdaBoostRegressor","BaggingRegressor","ExtraTreesRegressor","GradientBoostingRegressor","RandomForestRegressor"]

i=0

#subplot用
j=511

#对股价和各个因子进行拟合，并给出分数
plt.figure(figsize=(20,20))
while(i<len(name_list)):
    regr=function_list[i]
    regr.fit(Factors.loc[:,'ASSI':'JDQS20'],Factors.loc[:,'closePrice'])
    sc=regr.score(Factors.loc[:,'ASSI':'JDQS20'],Factors.loc[:,'closePrice'])
    #print("function %s score is %.8f"%(name_list[i],sc))
    
    predict=regr.predict(Factors.loc[:,'ASSI':'JDQS20'])
    Factors['predict']=predict
    plt.subplot(j)
    plt.plot(Factors.index,Factors.closePrice,label='closePrice')
    plt.plot(Factors.index,Factors.predict,label='predictPrice')
    plt.title("The score of function %s is %.8f"%(name_list[i],sc))
    plt.legend()
    i=i+1
    j=j+1
```

```python
#对相关情况排序，以便筛选相关系数绝对值最大的项
corrmat=Factors.corr()['closePrice'].sort_values()

#删除上面用于预测股价构建的衍生变量predict
corrmat=corrmat.drop("predict")

#选出因子中相关系数绝对值大于0.7的项
corrmat=corrmat.loc[(corrmat>0.7)|(corrmat<-0.7)]
cm = np.corrcoef(Factors[corrmat.index].values.T)
plt.figure(figsize=(18,16))
hm = sns.heatmap(cm, cbar=True, annot=True, square=True, fmt='.2f',yticklabels=corrmat.index.values, xticklabels=corrmat.index.values)
```
