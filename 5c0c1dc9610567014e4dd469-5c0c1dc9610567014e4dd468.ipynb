```python
import matplotlib.pyplot as plt
import numpy as np 
import pandas as pd
import statsmodels.api as sm
from statsmodels.graphics.tsaplots import plot_pacf,plot_acf

stockdata=DataAPI.MktAHCompdGet(beginDate=u'20060101',endDate=u'20161231')
stock_ticker_list=stockdata['tickerA']
stock_ticker_list=np.unique(stock_ticker_list)
stockdata=DataAPI.MktAHCompdGet(ticker=stock_ticker_list)
stockdata=stockdata.dropna(axis=0,how='any')
const=[]
l1_y=[]
l2_y=[]
l3_y=[]
l4_y=[]
l5_y=[]
l6_y=[]
count_ADF=0
for N in stock_ticker_list:
    stockdata_temp=stockdata[stockdata['tickerA']==N]
    stockpremium=stockdata_temp['closePriceA']/stockdata_temp['closePriceHA']
    stockpremium=stockpremium[::30]
    MAXLAGS = 6
    model = sm.tsa.AR(stockpremium)
    results = model.fit(MAXLAGS)
    const.append(results.params.const)
    l1_y.append(results.params[1])
    l2_y.append(results.params[2])
    l3_y.append(results.params[3])
    l4_y.append(results.params[4])
    l5_y.append(results.params[5])
    l6_y.append(results.params[6])
    plot_acf(stockpremium)
    plt.show()
    temp = np.array(stockpremium)
    t = sm.tsa.stattools.adfuller(temp)
    output=pd.DataFrame(index=['Test Statistic Value', "p-value", "Lags Used", "Number of Observations Used","Critical Value(1%)","Critical Value(5%)","Critical Value(10%)"],columns=['value'])
    output['value']['Test Statistic Value'] = t[0]
    output['value']['p-value'] = t[1]
    output['value']['Lags Used'] = t[2]
    output['value']['Number of Observations Used'] = t[3]
    output['value']['Critical Value(1%)'] = t[4]['1%']
    output['value']['Critical Value(5%)'] = t[4]['5%']
    output['value']['Critical Value(10%)'] = t[4]['10%']
    if t[1]<0.05:
        count_ADF+=1
        print(output)
        aic=sm.tsa.arma_order_select_ic(stockpremium,max_ar=12,max_ma=4,ic='aic')['aic_min_order']
        print(aic)
print(count_ADF)
    

#plt.hist(const,bins=15,normed=True)
#plt.show()
#plt.hist(l1_y,bins=15,normed=True)
#plt.show()
#plt.hist(l2_y,bins=15,normed=True)
#plt.show()
#plt.hist(l3_y,bins=15,normed=True)
#plt.show()
#plt.hist(l4_y,bins=15,normed=True)
#plt.show()
#plt.hist(l5_y,bins=15,normed=True)
#plt.show()
#plt.hist(l6_y,bins=15,normed=True)
#plt.show()
```
