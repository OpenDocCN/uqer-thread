# 机器学习选股模型

机器学习最近几年越加火热，其算法越来越成熟，也开始应用于各种应用中，如聊天机器人，谷歌翻译等等。很自然的，很多人尝试在投资中使用机器学习。目前主要有3类（本小菜自己分的）：
1. “激进者”开始尝试将机器学习完全替代研究员和基金经理， 使用机器学习进行选股或者择时，或者是进行机器人投顾（大部分应该是只是做基金的配置），这一类目前比较火热，但并没有大面积推广。
2. “保守派”完全不使用机器学习。主要是因为机器学习（特别是神经网络、深度学习等）容易给人"黑箱模型"和"过拟合"的印象。这种情况也的确存在，因为投资的一个重要原则就是“买自己懂的”，传统的方法强调策略的可解释性。如果对自己的模型或者策略无法深入了解，那么当模型失效的时候就是个灾难。这一类人占据了绝大多数。
3. 第三类应该是个人最为推崇的方式：传统+机器学习。俗话说万丈高楼平地起，传统的金融工程理论已经发展了几十年了，已经有非常坚实的理论基础，而且经历过实践的检验，不应该把这座宏伟的“大厦”完全推倒，而应该采取将机器学习嵌入传统的理论框架中。最近看到东方证券的这个研报就提供了很好的一个视角，但是原研报还是比较复杂，本文尝试使用最简单的粗暴的方式来实现。


## 1. 传统+机器学习方法

1. 首先是数据输入，有些使用把能搜集到的数据全部扔进机器学习算法中，企图通过更多的数据和更多的维度得到更好的结果。但是一般模型都有这样一个问题：“garbage in, garbage out”。数据的输入应该是经过精心挑选的。在这一步，使用传统的因子分析框架，对因子做IC检验，只有通过IC检验的因子才作为输入，这样对结果就会有个更好的解释。

2. 第二步是选择机器学习算法。机器学习的模型非常多，不存在最强的模型，每种模型都有[适用的范围](http://scikit-learn.org/stable/tutorial/machine_learning_map/)。（*好吧，个人觉得SVM和随机森林综合表现更好一点*）.并不是所有的机器学习算法都很“黑盒”，树型的机器学习算法就可以提供较好的解释性。这里主要选取随机森林，因为随机森林的逻辑和结果都非常直白、参数少，过拟合概率低，同时还具有非常强的样本外的预测能力。

## 2. 下面的模块主要是将训练数据落地

为了方便调参，需要将训练数据先落到csv文件中，后续的调试就会比较快了。



```python
# 定义引用的包和全局变量

import pandas as pd
import numpy as np
from CAL.PyCAL import *
cal = Calendar('China.SSE')
from functools import wraps
import time

factor_ls = ['LCAP','VOL20','ADX','ILLIQUIDITY','Volatility','JDQS20','REVS5','FY12P','PE','PB','ROE','ROA','GrossIncomeRatio']
```

```python
def fn_timer(function):
    """统计函数运行时间的装饰器函数"""
    @wraps(function)
    def function_timer(*args, **kwargs):
        t0 = time.time()
        result = function(*args, **kwargs)
        t1 = time.time()
        print ("Total time running %s: %s seconds" %
               (function.func_name, str(t1 - t0))
               )
        return result

    return function_timer

def get_halt_tickers(trading_day):
    """获取当日停牌的股票"""
    df = DataAPI.SecHaltGet(beginDate=trading_day,endDate=trading_day,listStatusCD="L,S",field=u"secID",pandas="1")
    return df['secID'].tolist()

def get_trading_day_list(begin_day_str, end_day_str):
    """获取交易日历"""
    cal_dates = DataAPI.TradeCalGet(exchangeCD=u"XSHG", beginDate=begin_day_str, endDate=end_day_str, field="calendarDate,isOpen")
    trading_days = cal_dates[cal_dates['isOpen']==1]["calendarDate"].tolist()
    return trading_days

def get_factor_data(trading_day):
    """ 读取因子数据"""
    all_field = 'secID,tradeDate,' + ','.join(factor_ls)
    return DataAPI.MktStockFactorsOneDayGet(tradeDate=trading_day,secID=u"",ticker=u"",field=all_field,pandas="1")

def get_factor_and_ret(begin_date, end_date):
    """获取因子数据和下期收益"""
    result = []
    trading_day_ls = get_trading_day_list(begin_date, end_date)
    for index, trading_day in enumerate(trading_day_ls):
        print trading_day
        if index < len(trading_day_ls) - 1:
            tomorrow = trading_day_ls[index+1]  # 获取下一个交易日
        else:
            break
        factor_df = get_factor_data(trading_day)
        ret_df = DataAPI.EquRetudGet(listStatusCD=u"L",secID=set_universe('A', tomorrow),ticker='',beginDate=tomorrow,endDate=tomorrow,isChgPctl=u"",field=u"secID,dailyReturnNoReinv",pandas="1")
        factor_df = factor_df.merge(ret_df, on='secID', how='inner')
        result.append(factor_df)
    return pd.concat(result) 
        

    #下面的代码将数据落成csv文件，
# df = get_factor_and_ret('20150101', '20161201')
# df.to_csv("factor_data/all_data.csv", index=False)

```

## 3. 回归的方法
1. 在每个调仓日（按周调仓），对前20个交易日的因子数据做回归，然后用回归系数
2. 取出昨日的因子数据作为输入，使用上一步得到的模型预测调仓日的收益
3. 将停牌股票剔除
4. 取出今日收益最大的100个股票
5. 将现有的仓位全部清空
6. 等量买入这100个股票

```python
from sklearn.ensemble import RandomForestRegressor

@fn_timer
def rf_predict(train_data, predict_date):
    train_data = train_data.fillna(0.0)
    X = train_data[factor_ls].values
    y = train_data['dailyReturnNoReinv'].values
    regressor = RandomForestRegressor(n_estimators=10, oob_score=True, min_samples_leaf=15)# 决策树回归
    regressor.fit(X, y)

    predict_data = get_factor_data(predict_date)
    predict_data = predict_data[~predict_data.secID.isin(get_halt_tickers(predict_date))]  # 滤除停牌的股票
    predict_data = predict_data.dropna()
    temp_x = predict_data[factor_ls].values
    ls = regressor.predict(temp_x)
    ret_se = pd.Series(ls, index=predict_data['secID'])
    ret_se = ret_se.nlargest(100) # 取预期收益最大的100个股票
    return ret_se.index.tolist() 
    
```
