# 等权重因子组合策略研究

本文尝试找一些长时间看来比较有效的因子并进行等权组合，就此选取一部分股票构成一个组合，就此形成一个完整的策略分析报告。
最终策略的风险指标如下：
|指标及分组|年化收益率|基准年化收益率|阿尔法|贝塔|夏普比率|波动率|信息比率|最大回撤|换手率|
|-|-|-|-|-|-|-|-|-|-|
|组合|16.20%|10.12%|6.55%|0.93|0.53|24.02%|0.65|40.13%|7.95|

下面是该策略的具体细节。

本文用到的因子主要有：

- 估值类因子：市盈率(PE)，市净率(PB)，市销率(PS)，市现率(PCF)
- 盈利类因子：净资产收益率(ROE)、总资产报酬率(ROA)、销售毛利率(GrossIncomeRatio)、销售净利率(NetProfitRatio)
- 动量反转因子：股票的5日收益(REVS5)，股票的10日收益(REVS10)、股票的20日收益(REVS20)
- 交投因子：5日平均换手率(VOL5)，20日平均换手率(VOL20)，60日平均换手率(VOL60)

首先，我们对于单个因子进行选股建仓调仓，并与benchmark进行比较。其后我们将一并展示其他因子的结果，对于每类因子选择一个最有效地因子。
而后，我们对于每类因子中最有效的因子进行等权组合，并由此进行选股与建仓，得到我们等权因子组合策略。

具体策略如下：

- 时间跨度设置为2013年1月1日至2017年12月31日，股票池为沪深300股票。
- 我们调仓的日期为每月第一个交易日。
- 按指标将300支成分进行排序，每50支股票为一档，并对每一档分别进行建仓。
- 调仓时股票的买入卖出价为当日收盘价。
- 等权重构建组合。
- 组合持有期内不调整组合。
- 组合构建时，买卖成本均为0.1%、买卖佣金均为0.1%，印花税为0.1%。

## 2.同类因子下因子的选择

对于同为估值类因子，上述策略的回测找到了对于PE因子最佳的指标分组。同样地，我们对于PB，PS，PCF也相应地找到最佳的指标分组，如下表所示：

|指标及分组|年化收益率|基准年化收益率|阿尔法|贝塔|夏普比率|波动率|信息比率|最大回撤|换手率|
|-|-|-|-|-|-|-|-|-|-|
|PE 0-50|14.97%|10.12%|4.9%|0.99|0.45|25.25%|0.64|42.22%|5.94|
|PB 0-50|17.94%|10.12%|7.89%|0.99|0.56|25.65%|0.84|43.34%|5.93|
|PS 0-50|16.08%|10.12%|5.81%|1.02|0.47|26.78%|0.62|49.22%|4.18|
|PCF 101-150|13.29%|10.12%|2.98%|1.03|0.37|26.32%|0.43|49.78%|11.44|

对于估值类因子，我们可以选择PB因子，并且选择因子值由小到大排列的前50个用于建仓。我们这里选择因子的标准主要是首先因子的组别顺序与收益有明显的正向或者负向相关性，其次，最佳组别的收益要有比较显著的alpha水平。

## 3.不同类因子的组合策略

仿照上面估值类因子的选择方式，我们分别对另外三类因子：盈利类，动量反转因子和交投因子进行因子选择和最佳分组的确定。得到以下的结果：

|指标及分组|年化收益率|基准年化收益率|阿尔法|贝塔|夏普比率|波动率|信息比率|最大回撤|换手率|
|-|-|-|-|-|-|-|-|-|-|
|PB 0-50|17.94%|10.12%|7.89%|0.99|0.56|25.65%|0.84|43.34%|5.93|
|ROE 250-300|11.76%|10.12%|4.06%|0.63|0.47|17.60%|0|26.28%|5.05|
|VOL60 0-50|14.12%|10.12%|4.85%|0.87|0.47|22.37%|0.41|37.84%|7.5|

这里没有动量类因子的原因是动量类因子值的大小关系与收益率没有显著地线性相关性，故在这里舍弃这类因子，只考虑以上三个因子。
现在我们将三个因子进行组合，得到新的等权重因子组合策略：

- 时间跨度设置为2013年1月1日至2017年12月31日，股票池为沪深300股票。
- 我们调仓的日期为每月第一个交易日。
- 分别计算每只股票的PB，1/ROE，VOL60指标，并进行标准化和归一化处理，将三个指标相加得到新的指标。
- 按新的指标将300支成分进行排序，每50支股票为一档，并对第一档即1-50进行建仓。
- 调仓时股票的买入卖出价为当日收盘价。
- 等权重构建组合。
- 组合持有期内不调整组合。
- 组合构建时，买卖成本均为0.1%、买卖佣金均为0.1%，印花税为0.1%。

通过回测我们可以看到，这个策略的alpha达到了6.6%，年化收益率达到了16.2%。具体指标如下：

|指标及分组|年化收益率|基准年化收益率|阿尔法|贝塔|夏普比率|波动率|信息比率|最大回撤|换手率|
|-|-|-|-|-|-|-|-|-|-|
|组合|16.20%|10.12%|6.55%|0.93|0.53|24.02%|0.65|40.13%|7.95|
