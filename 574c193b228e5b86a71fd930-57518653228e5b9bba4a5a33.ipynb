	以下部分来自于研报：
    
    价差偏离度因子本质上是一个相对意义上的反转因子，价差偏离度低，近期跑输其同类股票，股票相对处于低位，有向上回复的动力，有正的预期超额收益，价差偏离度越高，股票处于相对高位，后期有回调的压力。
    
    价差偏离度和传统的市值因子、估值因子相关性弱，通过因子分层后分组和Fama-Macbeth回归我们发现，传统的1个月反转和3个月反转可以被价差偏离度替代，而价差偏离度因子信息来源相对独立，不能被其他的常见因子所解释。

	由此可见，价差偏离度这个因子，主要作用是捕捉股票相对其同类型（板块）股票的高估低估程度。低估的股票之后会补涨，高估的股票之后有补跌的趋势。

这里我们介绍一下价差偏离度的计算方法：

股票$i$的价差偏离度的定义按如下三步：

(1) 每月月底在全市场搜索与股票 $i$ 距离最近（相似度最高）的 $N$ 只股票等权构建该股票的特征组合，特征组合的净值价格我们称之为参考价格($ReferencePrice_{i,t}$)

(2)计算股票 $i$ 的价格和参考价格的对数价差（$PriceSpread_{i,t}$):
$$PriceSpread_{i,t}=ln(StockPrice_{i,t})-ln(ReferencePrice_{i,t})$$

(3)经过去 $60$ 日均值、标准差调整后的当前对数价差 $PriceSpread_{i,t}$ ，我们称之为价差偏离度。
$$ SpreadBias_{i,t}=\frac{PriceSpread_{i,t}-mean(PriceSpread_{i,t})}{std(PriceSpread_{i,t})} $$

关于股票间的距离，我们可以通过股票过去的日涨跌幅的相关性来变相度量股票间的距离
     
     股票间的距离=1-两个股票过去250个交易日涨跌幅的pearson相关系数

关于两个参数：股票组合的容量N和计算价差偏离度的时间窗口K的选择，原研报都有详细解释，这里就不赘述了。

```python
from CAL.PyCAL import *
import pandas as pd
from scipy.stats.stats import pearsonr
import numpy as np
start = '2011-08-01'
end = '2016-04-01'
limit='2016-06-06'
universe = set_universe('SH50')

#Obtaining the date of the month end
data=DataAPI.TradeCalGet(exchangeCD=u"XSHG",beginDate= start,endDate= limit,field=['calendarDate','isMonthEnd'],pandas='1')
data=data[data['isMonthEnd']==1]
monthend=data['calendarDate'].values.tolist()

cal = Calendar('China.SSE')

#compute the distance between two given stks on a given date
def distance(stk1,stk2,Date,peri=250):
    if stk1==stk2:
        return 0.0
    else:
        cal = Calendar('China.SSE')
        trueperiod=Period('-'+str(peri)+'B')
        startdate=cal.advanceDate(Date,trueperiod)
        startdate =startdate.toDateTime()
        ss=startdate.strftime('%Y-%m-%d')
        df1=DataAPI.MktEqudAdjGet(secID=stk1,field=['tradeDate','closePrice'],beginDate=ss,endDate=Date,pandas='1').set_index('tradeDate')
        df1.rename(columns={'closePrice':stk1},inplace=True)
        df1[stk1]=df1[stk1].pct_change()
        df2=DataAPI.MktEqudAdjGet(secID=stk2,field=['tradeDate','closePrice'],beginDate=ss,endDate=Date,pandas='1').set_index('tradeDate')
        df2.rename(columns={'closePrice':stk2},inplace=True)
        df2[stk2]=df2[stk2].pct_change()
        df1=df1.merge(df2,left_index=True,right_index=True)
        df1.dropna(inplace=True)
        s=df1[stk1].corr(df1[stk2],method='pearson')
        dis=1.0-s
    return dis

#Find the target list
def targetlist(stk,Date,N=10):
    target={}
    for i in universe:
        diss=distance(stk,i,Date)
        target[i]=diss
    dic= sorted(target.iteritems(), key=lambda d:d[1])    #因为字典不支持排序，我们先把字典转换成二元数组进行排序
    dics=[item[0] for item in dic ]
    return (dics[0:N] if stk not in dics else dics[1:N+1])

#compute the reference price
def reference_price(target,Date):
    df=DataAPI.MktEqudGet(secID=target,tradeDate=Date,field=u"tradeDate,closePrice,secID",pandas="1").set_index("tradeDate")   
    price=df['closePrice'].mean()
    return np.log(price)

#compute the pricebias of stk on a given date
def pricebias(stk,Date,K=60):
    #Acquiring the trading date information
    cal = Calendar('China.SSE')
    trueperiod=Period('-'+str(K)+'B')
    startdate=cal.advanceDate(Date,trueperiod)
    startdate =startdate.toDateTime()
    ss=startdate.strftime('%Y-%m-%d')
    #
    df=DataAPI.MktEqudGet(secID=stk,beginDate=ss,endDate=Date,field=u"tradeDate,closePrice",pandas="1").set_index("tradeDate")
    df['closePrice']=np.log(df['closePrice'])
    pl=targetlist(stk,Date)
    for i in xrange(len(df.index)):
        for j in xrange(len(monthend)):
            if df.index[i]>=monthend[j] and df.index[i]< monthend[j+1]:#因为我们是每月换仓，这里我们认为一个月内和某股票距离最相近的10只股票不会变
                df['closePrice'][i]=df['closePrice'][i]-(reference_price(pl,monthend[j]))
                break
    return float((df['closePrice'][-1]-df['closePrice'].mean())/df['closePrice'].std())

```

以上函数便完成了对于单个股票在给定日期价差偏离度的计算，利用这些函数我们来试着获取一下 '2016-04-01' 这天 沪深300 中10只价差偏离度最小的股票。 

```python
end = '2016-04-01'
universe=set_universe('HS300')
import datetime
starttime = datetime.datetime.now()

def oneday(target_list,Date,N=10):
    mm={stk:pricebias(stk,Date) for stk in target_list}
    dicc= sorted(mm.iteritems(), key=lambda d:d[1])
    portfolio=[item[0] for item in dicc ]
    return portfolio[0:N]

bbb=oneday(universe,end)
print bbb
endtime = datetime.datetime.now()
print 'The running time is %d s.' %((endtime - starttime).seconds)  
```

我们成功得到了16年4月1日沪深300中价差偏离度最小的十只股票，只不过这运行的时间有点やばいじゃないか？

策略：我们的策略很简单，每月月初计算上月底市场上各股票的价差偏离度，买入价差偏离度最小的10只，并持有一个月。每月换仓。

回测结果不是很好，这也是在意料之中，因为本来应该在全市场中搜寻股票，而本文主要的目的是向大家介绍这个因子，为了节省运行时间，universe选取了上证50，导致价差偏离度的计算出了很大的偏差（从图中也能看出，股票组合的走势几乎和上证50完全相同）。实际上，如果选取沪深300或者整个A股的话（universe选的越大计算出来的价差偏离度也就越准确），回测的结果会好很多，当然，运行时间也会成倍的增加，所以最好做一个数据库把已经跑出来的结果存起来，之后直接调用，这样运行速度会快很多。
