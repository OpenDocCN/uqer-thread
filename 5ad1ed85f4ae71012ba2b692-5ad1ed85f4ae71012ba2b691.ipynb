# 机器学习之Logistic因子选股

# **前言**
由于因子选股简单有效，历来是券商们研究的重点，本文采用Logistic方法，利用Uqer的数据进行量化分析，建立量化选个的基本分析框架


# **数据准备**
--------
本文选取了优矿普通用户所能够采用的所有因子，提取股票每个月末的因子暴露作为训练输入特征。读者也可以选取自己感兴趣的因子作为基础因子。
特征进行缺失值处理，标准化等处理，该处理后的数据作为模型输入特征。同时在每个月末截面期，选取下月收益排名前30%的股票作为正例（𝑦=1），后30%的股票作为负例（𝑦=−1）。利用sklearn进行分类预测。读者可根据自己常用特征处理习惯进行处理。

**提醒**：
	本文的数据处理较为简单，实际中可根据自己需要处理。另，每月最后一天可能是非交易日，这里并没有对交易日过滤，这会使得部分股票持仓周期为2个月。

```python
from __future__ import print_function
import datetime
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression,RandomizedLogisticRegression
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import roc_auc_score,roc_curve
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#日期选择，选择每月的最后一天
data_end=[]
for i in range(2007,2018):
    for j in xrange(1, 13):
        if 12 == j:
            dt_end = (datetime.datetime(i, 12, 31)).strftime("%Y%m%d")
        else:
            dt_end = (datetime.datetime(i, j+1, 1) - datetime.timedelta(days = 1)).strftime("%Y%m%d")
        data_end.append(dt_end)

    
#从Uqer导入数据
data0=pd.DataFrame()
for i in data_end:
    data1=DataAPI.MktStockFactorsOneDayGet(secID=set_universe('HS300'),tradeDate=i,ticker='',field=u"",pandas="1").set_index(['secID']) 
    data2=DataAPI.MktEqudGet(secID=set_universe('HS300'),tradeDate=i,field=u"secID,closePrice",pandas="1").set_index(['secID']) 
    data1=pd.concat([data1,data2],axis=1)
    data1['stock']=data1.index
    data1=data1.set_index(['tradeDate'])
    data0=pd.concat([data0,data1])
#data0.to_csv('Data300.csv',encoding='utf-8')


#预处理
print ('预处理')
data0['tradeDate']=data0.index
data0=data0.reset_index(drop=True)
data0=data0.loc[data0['tradeDate'].dropna().index]
#按月分离交易日期，方便回测            
data0['y_m'] = data0.tradeDate.apply(lambda x: str(x[0:7]))   

#收益率分类（1 上涨前30%，0下跌前30%）
data_price=data0[['stock','closePrice']]
ret=pd.DataFrame()
for i in set(data0['stock']):
    d=data_price[data_price['stock']==i]
    d_ret=(d.shift(-1).iloc[:-1,1]-d.iloc[:-1,1])*1.0/d.iloc[:-1,1]
    ret=pd.concat([ret,d_ret])
del data_price,d,d_ret
ret=ret.fillna(0)
q07=ret.quantile(0.7)[0]
q03=ret.quantile(0.3)[0]
for i in ret.index:
    if ret.loc[i][0]>q07:
        ret.loc[i]=1
    elif ret.loc[i][0]<q03:
        ret.loc[i]=0
    else:
        ret.loc[i]=-1
ret.columns=['rets']


#因子处理
data_factor=data0.drop(['closePrice','ticker'],axis=1)
data_factor=data_factor.loc[ret.index]
factors=data_factor.columns[1:-2]

#缺失值处理
print ('缺失值处理')
data_factor[factors]=data_factor[factors].fillna(0)

#除去极端值
#暂略

#因子归一化
print ('因子归一化')
data_factor[factors]=(data_factor[factors]-data_factor[factors].mean())*1.0/data_factor[factors].std()

#数据合并
data_factor=pd.concat([data_factor,ret],axis=1)
df=data_factor
df.head()
```

# **模型训练**
-----
为了能让模型及时抓取到市场的变化，我们采用了滚动回测方法。模型训练区间为20070101至20171231，按月份划分（前12个月训练，本月测试），因此需要不断重复训练。
get_train_val_test_data()：数据集划分函数
format_feature_label()：特征划分函数
pipeline():模型处理，包括因子选择，格点搜索等等

利用格点搜索做交叉验证，选择最好的模型，在测试集测试

注意：
	部分函数的编写参考Uqer社区的 华泰证券 《人工智能选股之 Boosting 模型》这篇文章

```python
def get_train_val_test_data(date):
        #计算前12个月，采用365天
        d_12=(datetime.datetime.strptime(date, "%Y-%m") - datetime.timedelta(days=365)).strftime('%Y-%m')
        train_df = df[(df['y_m']>=d_12) & (df['y_m']<date)]
        #train_val_df = train_val_df.sample(frac=1).reset_index(drop=True)   
        #拆分训练集、验证集                                 
        test_df = df[df['y_m']==date]   
        return train_df,test_df

def format_feature_label(origin_df, is_filter=True):
    
    if is_filter:
        origin_df = origin_df[origin_df['rets']!=-1]
    df_test= origin_df                
    feature = origin_df[factors]
    label = origin_df['rets']

    return feature, label,df_test

def pipeline():
    LR_model_list = []
    accuracy=[]
    final=pd.DataFrame()
    for date in df['y_m'].drop_duplicates():
        if date>'2008-01':
            print('training model for %s' % date)
            train_df,test_df = get_train_val_test_data(date)
            train_feature, train_label,df_train = format_feature_label(train_df)       
            test_feature, test_label,df_test = format_feature_label(test_df)
            
            #训练集           
            #特征选择
            rlr=RandomizedLogisticRegression()
            rlr.fit(train_feature,train_label)         
            train_feature1=train_feature[train_feature.columns[rlr.get_support()]]       
            #建立模型
            #print('LR model')                                                
            lr=LogisticRegression()           
            #格点搜索
            #print('GridSearchCV')   
            params={'C':[0.1,1,10,100]}
            clf=GridSearchCV(lr,param_grid=params,n_jobs=-1)
            clf.fit(train_feature1,train_label)
            clf_best=clf.best_estimator_
            clf_best.fit(train_feature1,train_label)
        
            #验证集        
            test_feature1=test_feature[test_feature.columns[rlr.get_support()]]             
            predict_prob=clf_best.predict_proba(test_feature1)[:,1]                                                
            #clf_best.score(val_feature1,val_label)
            print('accuracy : %s '%(clf_best.score(test_feature1,test_label)))
            accuracy.append(clf_best.score(test_feature1,test_label))
            LR_model_list.append(clf_best)
            df_test.loc[:,'factor'] = predict_prob
            df_test = df_test.loc[:, ['stock', 'tradeDate', 'rets', 'factor']] 
            final=pd.concat([final,df_test])
    return LR_model_list,final,accuracy
```

# **结果分析**
------
对准确率画图，这里需要注意的是，在训练中对收益率进行了划分前30%和后30%，相当于略去了中间40%的数据，但是在测试集中我们无法知道哪些股票未来上涨高，哪些上涨低，所以无法删去数据，这就使得测试的时候，即使预测值为1，也无法保证一定是所有股票中上涨排名前30%的股票，不过在券商 华泰证券《人工智能选股》系列研报中均采用了此种方法，本文姑且认为这个方法可行

```python
#运行模型
LR_model_list,final,accuracy = pipeline()

#准确率分析
accuracy=pd.DataFrame()
date_test=pd.DataFrame(data_end)
#对各阶段的准去率作图
print('accuracy=%f'% (accuracy.mean()))
plt.plot(accuracy,label='accuracy')
```

# **结果分析**
利用简单的Logistic回归，可取得不错的收益率，但是需要注意几个问题：
1、月末的非交易日并未处理，导致部分股票的持仓超过1个月
2、HS300作为基准每月买入的股票数不够多，可以采用ZZ500或者全A股
3、因子筛选方面，从240个因子中采用随机逻辑斯提回归筛选，并没有从经济角度筛选因子，有失偏颇
4、资金是1千万，HS300中满足条件的股票不多，所以每次买卖量会较大，这必然会产生**滑点**，这个是必须考虑的
