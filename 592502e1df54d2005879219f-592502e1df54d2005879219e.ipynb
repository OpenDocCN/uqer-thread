Matplotlib库由各种可视化类构成，内部结构复杂，受Matlab启发。
matplotlib.pyplot是绘制各类可视化图形的命令子库，相当于快捷方式。
使用方式：
import matplotlib.pyplot as plt

```python
#直接绘制一个序列，横坐标为索引号
import matplotlib.pyplot as plt
plt.plot([3,1,4,5,2])
plt.ylabel("Grade")
plt.savefig('test',dpi=600)
plt.show()
#plt.plot()只有一个输入列表或数组时，参数被当作Y轴，X轴以索引自动生成
#plt.savefig()将输出图形存储为文件，默认PNG格式，可以通过dpi修改输出质量
```

```python
#指定X轴、Y轴值，并指定坐标轴范围
import matplotlib.pyplot as plt
plt.plot([0,2,4,6,8],[3,1,4,5,2])
plt.ylabel("Grade")
plt.axis([-1,10,0,6])
plt.show()
#plt.plot(x,y)当有两个以上参数时，按照X轴和Y轴顺序绘制数据点
```

```python
#在全局绘图区域中创建一个分区体系，并定位到一个子绘图区域
import numpy as np
import matplotlib.pyplot as plt
def f(t):
    return np.exp(-t)*np.cos(2*np.pi*t)

a = np.arange(0.0,5.0,0.02)

plt.subplot(2,1,1)
plt.plot(a,f(a))

plt.subplot(2,1,2)
plt.plot(a,np.cos(2*np.pi*a),'r--')  #原始出处a写成t2，应该为a.
plt.show()
```

plt.plot(x, y, format_string, **kwargs)
x : X轴数据，列表或数组，可选
y : Y轴数据，列表或数组
format_string: 控制曲线的格式字符串，可选
**kwargs : 第二组或更多(x,y,format_string)
当绘制多条曲线时，各条曲线的x不能省略

```python
import numpy as np
import matplotlib.pyplot as plt
a = np.arange(10)
plt.plot(a,a*1.5,a,a*2.5,a,a*3.5,a,a*4.5)
plt.show()
```

format_string: 控制曲线的格式字符串，可选
由颜色字符、 风格字符和标记字符组成

如颜色字符对应关系如下：
颜色字符 说明 颜色字符 说明
'b' 蓝色 'm' 洋红色 magenta
'g' 绿色 'y' 黄色
'r' 红色 'k' 黑色
'c' 青绿色 cyan 'w' 白色
'#008000' RGB某颜色 '0.8' 灰度值字符串

另外的风格字符、标记字符请参考原始文档。

针对上面的例子，可以对曲线控制如下：

```python
import numpy as np
import matplotlib.pyplot as plt
a = np.arange(10)
plt.plot(a,a*1.5,'go-',a,a*2.5,'rx',a,a*3.5,'*',a,a*4.5,'b-.')
plt.show()
```

pyplot的中文显示方法：
一、使用rcParams修改字体实现
二、在有中文输出的地方，增加一个属性：fontproperties

```python
#本文第1个例子绘制的Y坐标轴是英文，若是中文将不支持
import matplotlib.pyplot as plt
import matplotlib
matplotlib.RcParams['font.family']='STSong'
plt.plot([3,1,4,5,2])
plt.ylabel("纵轴")
plt.savefig('test',dpi=600)
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.plot([3,1,4,5,2])
plt.ylabel('纵轴',fontproperties='SimHei',fontsize=20)
plt.savefig('test',dpi=600)
plt.show()
```

#pyplot的文本显示函数
函数           说明
plt.xlabel() 对X轴增加文本标签
plt.ylabel() 对Y轴增加文本标签
plt.title() 对图形整体增加文本标签
plt.text() 在任意位置增加文本
plt.annotate() 在图形中增加带箭头的注解
