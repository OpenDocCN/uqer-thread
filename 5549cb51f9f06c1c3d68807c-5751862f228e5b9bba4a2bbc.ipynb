通过数据API: ``NewsHeatIndexGet`` 获取某只股票对应的**新闻热度**。

```python
res = DataAPI.NewsHeatIndexGet(secID = '600000.XSHG',field=['secID', 'newsPublishDate', 'heatIndex'])
res.tail()
```

上面的API调用，获得了最近的浦发银行的每日新闻情感：

- ``secID`` 证券代码
- ``newsPublishDate`` 交易日
- ``NewsHeatIndexGet`` 新闻热度指数，证券当天关联新闻数量占当天关联新闻总量的百分比(%）


3. 策略实现
----------------------
***

- **投资域** ：沪深300成分股
- **业绩基准** ：沪深300指数
