```python
#Author: JasonYichuan
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('white')

# LOB 画图
def LOBplot(ticker):
    import matplotlib.gridspec as gridspec
    if ticker[0] == '6':
        securityID = ticker + '.XSHG'
    else:
        securityID = ticker + '.XSHE'
    df = DataAPI.MktTickRTIntraDayGet(securityID=securityID,startTime=u"09:30",endTime=u"14:56")
    #if ticker[0] != '6':
    #    df = df.ix[:len(df)-23].reset_index()
    spread = 0.01
    numDf = len(df)
    maxPrice = max(round(df.askPrice5.max(),2),round(df.askPrice4.max(),2))
    minPrice = max(round(df.bidPrice5.min(),2),round(df.bidPrice4.min(),2))
    numPrice = int((maxPrice - minPrice) / spread + 1)
    #print(maxPrice,minPrice,numPrice)
    priceRange = np.linspace(minPrice, maxPrice, numPrice)
    timeRange = df.dataTime
    hfData = np.zeros([numPrice,numDf])
    for i in range(numDf):
        hfData[int((round(df.askPrice1.ix[i],2) - minPrice)/spread),i] = df.askVolume1.ix[i]
        hfData[int((round(df.askPrice2.ix[i],2) - minPrice)/spread),i] = df.askVolume2.ix[i]
        hfData[int((round(df.askPrice3.ix[i],2) - minPrice)/spread),i] = df.askVolume3.ix[i]
        hfData[int((round(df.askPrice4.ix[i],2) - minPrice)/spread),i] = df.askVolume4.ix[i]
        hfData[int((round(df.askPrice5.ix[i],2) - minPrice)/spread),i] = df.askVolume5.ix[i]
        hfData[int((round(df.bidPrice1.ix[i],2) - minPrice)/spread),i] = df.bidVolume1.ix[i]
        hfData[int((round(df.bidPrice2.ix[i],2) - minPrice)/spread),i] = df.bidVolume2.ix[i]
        hfData[int((round(df.bidPrice3.ix[i],2) - minPrice)/spread),i] = df.bidVolume3.ix[i]
        hfData[int((round(df.bidPrice4.ix[i],2) - minPrice)/spread),i] = df.bidVolume4.ix[i]
        hfData[int((round(df.bidPrice5.ix[i],2) - minPrice)/spread),i] = df.bidVolume5.ix[i]
    data = pd.DataFrame(hfData,columns = timeRange).set_index(priceRange)
    data.columns.name = 'Time'
    data.index.name = 'Price'
    #print(data)
    #heat map
    plt.figure(figsize=(18,10))
    gs = gridspec.GridSpec(2, 1,height_ratios=[4,1])
    ax1 = plt.subplot(gs[0])  
    plt.pcolor(np.arange(len(df)),priceRange,hfData,cmap='spectral',alpha=0.7)
    plt.plot(np.arange(len(df)),df.lastPrice+spread,color='w',label='Last Price')
    plt.plot(np.arange(len(df)),df.bidPrice1+spread,color='r',label='Best Bid Price')
    plt.plot(np.arange(len(df)),df.askPrice1+spread,color='g',label='Best Ask Price')
    #plt.legend()
    plt.xlim(0,len(df))
    plt.ylim(minPrice,maxPrice)
    xvalue = np.arange(0,len(df),int(len(df)/16))
    xtick = [t for t in df.dataTime]
    xtick = list(map(lambda t: xtick[t], xvalue))
    plt.xticks(xvalue,xtick)
    plt.xlabel('Time')
    plt.ylabel('Price')
    plt.title(str(df.ticker.ix[0]))
    plt.colorbar(orientation='horizontal', pad=0.06)
    
    #volume
    ax2 = plt.subplot(gs[1])
    volume = df.volume - df.volume.shift()
    plt.plot(np.arange(len(df)),volume,label='VOLUME',color='k')
    plt.xlim(0,len(df))
    plt.xticks(xvalue,xtick)
    plt.legend()
    plt.grid()
```
