# 基金选股择时能力_T-M、H-M、C-L模型计量

选股择时能力
   评价基金产品其选股，择时能力。经典量化模型：C-L模型，H-M模型，T-M模型，都是对基金收益率和市场超额收益率之间进行线性回归分析。
    在上述三个模型下，α表示的是基金的超额收益率，度量投资组合获取超额收益的能力，一般会归因于基金经理的选股能力；β2衡量的是基金经理的择时能力。三个模型和计量公式汇总如下：

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
```

```python
#取数据：无风险收益率、市场组合收益率和要评价的基金收益率

beginDate="2012-04-16"
endDate="2018-04-16"

#无风险收益率，用银行间3个月Shibor
rf=DataAPI.MktIborGet(secID=u"Shibor3M.IRCN",ticker=u"",tradeDate=u"",beginDate=beginDate,endDate=endDate,currency=u"",field=u"ticker,tradeDate,rate",pandas="1").pivot_table(index='tradeDate', columns='ticker', values='rate').bfill()
rf=rf/100;

#市场组合收益率数据，用沪深300指数代替
m=DataAPI.MktIdxdGet(tradeDate=u"",indexID=u"",ticker=u"000300",beginDate=beginDate,endDate=endDate,exchangeCD=u"XSHE,XSHG",field=u"indexID,tradeDate,closeIndex",pandas="1").pivot_table(index='tradeDate', columns='indexID', values='closeIndex').bfill()
r_m=m.pct_change().dropna()

#基金收益率数据,举例：博时沪深300指数A，050002，任意选的
r_fund=DataAPI.FundNavGet(dataDate=u"",secID=u"",ticker='050002',beginDate=beginDate,endDate=endDate,field=u"ticker,endDate,ADJUST_NAV",pandas="1").pivot_table(index='endDate', columns='ticker',values='ADJUST_NAV').bfill()   
r_fund=r_fund.pct_change().dropna()
```

```python
#有收益率数据，线性回归函数
def regression(x1,x2, y):
    x1, x2 = x1.reshape(len(x1), 1), x2.reshape(len(x2),1)
    c = np.ones((len(x1),1))
    X = np.hstack((c,x1,x2))
    res = (sm.OLS(y,X)).fit()  
    return res.summary()
```

```python
#数据整理
data1 = pd.concat([r_fund,rf],axis=1).dropna()
data2=pd.concat([data1,r_m],axis=1).dropna()
data2.columns = ['r_fund', 'r_f','r_m']
data2.head()
```

```python
 #T-M模型
y=data2['r_fund']-data2['r_f']
x1=data2['r_m']-data2['r_f']
x2=x1**2
regression(x1, x2, y)
```

```python
d = []                       #H-M模型
for i in range(len(data2['r_m'])):
    if data2['r_m'][i]>data2['r_f'][i]:
        d.append(1)
    else:
        d.append(0)
d = np.array(d)
y=data2['r_fund']-data2['r_f']
x1=data2['r_m']-data2['r_f']
x2=d*(x1**2 )      
        
res_HM = regression(x1, x2, y)
res_HM 
```

```python
x1,x2 = [], []               #C-L模型
for i in range(len(data2['r_m'])):
    if data2['r_m'][i]>data2['r_f'][i]:
        x1.append(data2['r_m'][i]-data2['r_f'][i])
        x2.append(0)
    else:
        x1.append(0)
        x2.append(data2['r_m'][i]-data2['r_f'][i])

x1,x2 = np.array(x1), np.array(x2)
y = data2['r_fund']-data2['r_f']
res_CL = regression(x1, x2, y)
res_CL
```
