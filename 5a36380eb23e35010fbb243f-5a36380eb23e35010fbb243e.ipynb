# 时间函数的处理datetime和time.time和Date

Python时间相关有3个库:
datetime
	now=datetime.datetime.now(),
    date=dattime.date(now)
    timedelta:可以进行日、时、分、秒的加减。
time
	time.time
    这个是时间戳，可以转为元组。在DateFrame可能会用上。是一个64位浮点数据可以和datetime进行转化。
CAL.Date
	优矿自带的时间函数（日期），可以进行年、月、日的加减。
    可以和datetime进行格式转化。

```python
from CAL.PyCAL import *
from datetime import datetime
# 用CAL.Date计算日期
end=Date.todaysDate()
start=end-Period('1D')
# Date转为datetime
end=end.toDateTime()
start=start.toDateTime()
# 取得日期去除时间
end=datetime.date(end)
start=datetime.date(start)

endStr=end.strftime("%Y%m%d")
startStr=start.strftime("%Y%m%d")
print endStr,startStr
```
