使用方法：直接修改start和end的时间为你想获取数据的时间运行，建议每两年运行一次，不然内存可能不够，运行完后在Data里会有个excel，可以下载。用的时候读取excel，第一列是tradedate+time的字符串（如2009-01-05 09:33），第二列是每分钟收盘价。我这里用API获取了每天的收盘价补到了每天的15:00，因为“策略”里不给每天的收盘价。根据优矿规则9:30的数据是集合竞价数据。这是获取沪深300指数的。

如果要获取别的指数，在benchmark里改一下，且在DataAPI那里的IndexID改一下就行了，这是获取上证指数的。

这是获取个股的，注意每次只能一只股;每天15:00只有收盘价、成交量、成交额（算出来的），每天最后一分钟的最高/低价是真没办法获取到，毕竟高频数据。。
