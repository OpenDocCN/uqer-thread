## 3.如何画趋势图（菜鸟心得-1）
***
 最近在研究分级基金套利的问题，需要画走势图，因此研究了一下python画图的功能。本来以为需要从matplotlib研究起，但是后来在社区里看到 @李自龙 写的一个简短的例子，让我茅塞顿开。原来dataframe就有绘图功能，不用从头学习matlib。
 
 所以接下来绘图的问题就转化为，如何构建一个符合自己需要的data frame了。
 
 在构建这个data frame的过程中，我走了一些弯路，所幸得到了社区高手的指点，最后还是绕回来了。 
 
 我的思路，都以注释的方式，体现在代码里了。 

```python
from CAL.PyCAL import *  #为了用font，必须引入这个库
rootFund=DataAPI.FundNavGet(dataDate=u"",secID=u"",ticker=u"161022",beginDate=u"20150101",endDate=u"20151231",field=u"endDate,NAV",pandas="1") #获取母基金净值
subFund=DataAPI.MktFunddGet(tradeDate=u"",secID=u"",ticker=u"150152,150153",beginDate=u"20150101",endDate=u"20151231",field=u"tradeDate,secID,openPrice",pandas="1") #为提高效率，一次性获取A、B类基金的开盘价
#为了方便计算，需要把A类、B类基金的开盘价，并排展示，所以重新构建了一个data frame
df=pd.merge(subFund[subFund.secID == '150152.XSHE'],subFund[subFund.secID == '150153.XSHE'],on='tradeDate')

#为母基金查询结果集，增加2列。一列是A、B基金合并价格，一列是母基金相对子基金的溢价
#新增列非常简单，直接指定一个之前不存在的列，赋上值就可以了。操作太方便了，超出预期，我就是在这里走了弯路。
#感谢 @taotao.li 的指点
rootFund['subFundTotal']=(df['openPrice_x'] + df['openPrice_y'])/2
rootFund['Gap']=rootFund['subFundTotal'] - rootFund['NAV']

#开始画图。由于RootFund数据集中有好多列，所以画图时，需要告诉python,用哪一列做横坐标，哪一列做纵坐标
ax = rootFund.plot(figsize=(10,5),x=u'endDate',y=u'Gap')
patches, labels = ax.get_legend_handles_labels()
#labels是图例说明的意思。这里需要显示的告知，说明里写什么内容，否则就是None
labels=[u'溢价']
ax.legend(patches, labels, loc='best', prop = font)  # 使用font
ax.set_title(u'2015创业板分级基金走势图', fontproperties = font, fontsize = 16)  # 使用font
ax.set_xlabel(u'日期', fontproperties = font, fontsize = 16)  # 使用font 
```

## 4.如何重设索引或者使索引归0
***
 首先当然是，为什么要重设索引？
 
 pandas支持两个不同索引的data frame进行计算，但是计算的结果，是通过匹配index来实现的，结果往往不是我们想要的。
 
 在这种情况下 ，只有将两个data frame的索引建成一样，才方便计算。
 
 接下来，结合一个具体的案例，来解释一下，为什么，以及如何重设索引

```python
#获取母基金净值
rootFund=DataAPI.FundNavGet(dataDate=u"",secID=u"",ticker=u"161022",beginDate=u"20150101",endDate=u"20151231",field=u"endDate,NAV",pandas="1") 
#获取子基金净值
subFund=DataAPI.MktFunddGet(tradeDate=u"",secID=u"",ticker=u"150152,150153",beginDate=u"20150101",endDate=u"20150131",field=u"tradeDate,secID,openPrice",pandas="1")
#把row变成列。
df=pd.merge(subFund[subFund.secID == '150152.XSHE'],subFund[subFund.secID == '150153.XSHE'],on='tradeDate') 
#过滤前两天的子基记录
df=df[df.tradeDate>'2015-01-06']
#此时的索引是从2开始编的

#为啥要过滤呢？因为在做分级基金套利的时候，T日申购的基金，在T+2日可以拆成A和B然后卖出。所以我需要将T日母基金净值与T+2日子基金价格放在同一行
#相当于excel表里，子基金的数据上移2行的效果

#将老索引删除，且覆盖df对象。如果没有inplace，那么就需要把删除索引后的数据赋给另外一个变量。
df.reset_index(drop=True,inplace=True)
#增加一个计算列，而且看得出，此时索引又从0开始编了。
df.assign(subTotal = lambda x: (x.openPrice_x+x.openPrice_y)/2)
```
