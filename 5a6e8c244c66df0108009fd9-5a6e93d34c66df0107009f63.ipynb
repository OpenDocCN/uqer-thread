```python
import numpy as np
import pandas as pd
from sklearn.decomposition import KernelPCA
import matplotlib.pyplot as plt
import matplotlib as mpl
```

读取上证50指数及其成分股的收盘价（closePrice）

```python
symbols = ['600000','600016','600019','600028','600029','600030','600036','600048','600050','600104','600111','600309','600340','600518','600519','600547','600606','600837','600887','600919','600958','600999','601006','601088','601166','601169','601186','601211','601229','601288','601318','601328','601336','601390','601398','601601','601628','601668','601669','601688','601766','601800','601818','601857','601878','601881','601985','601988','601989','603993']
startdate = '20160101'
enddate = '20180101'
```

```python
data = pd.DataFrame()
data =         DataAPI.MktEqudGet(ticker=symbols,beginDate=startdate,endDate=enddate,field=u"ticker,secShortName,tradeDate,closePrice",pandas="1")
```

```python
data = pd.crosstab(data['tradeDate'],data['secShortName'], values = data['closePrice'], aggfunc = sum)
data = data.fillna(method='bfill')    # 用后一个数值填充缺省值
data.head()
```

```python
sz50 = DataAPI.MktIdxdGet(beginDate=startdate,endDate=enddate,indexID=u"000016.ZICN",ticker=u"",field=u"indexID,tradeDate,closeIndex",pandas="1")
```

```python
sz50 = pd.crosstab(sz50['tradeDate'],sz50['indexID'],values=sz50['closeIndex'],aggfunc=sum)
# sz50 = ss50.loc[data.index]     # 根据data的inde筛选数据
sz50.head()
```

PCA通常处理规范化的数据集，这里将数据集标准正态化

```python
scale_function = lambda x: (x-x.mean())/x.std()        # 标准正态化
```

```python
pca = KernelPCA().fit(data.apply(scale_function))
```

```python
len(pca.lambdas_)
```

```python
# 观察前10个成分
pca.lambdas_[:10]
```

```python
# 前5个成分的总占比
get_we = lambda x: x / x.sum()
get_we(pca.lambdas_)[:5].sum()
```

构造PCA指数

首先，我们建立只包含第一个成分的PCA指数

```python
pca = KernelPCA(n_components=1).fit(data.apply(scale_function))
sz50['PCA-1'] = pca.transform(-data)
```

```python
fig = sz50.apply(scale_function).plot(figsize=(12,6))
```

再建立一个包含5个成分的PCA指数，看看是否能改善结果

```python
pca = KernelPCA(n_components=5).fit(data.apply(scale_function))
pca_components = pca.transform(-data)
weights = get_we(pca.lambdas_)
sz50['PCA-5'] = np.dot(pca_components,weights)
```

```python
fig = sz50.apply(scale_function).plot(figsize=(12,6))
```

通过散点图检查上证50指数和PCA指数之间的关系

```python
mpl_dates = mpl.dates.date2num(list(pd.to_datetime(data.index)))  # 将index的日期转换成matplotlib兼容的格式
```

```python
sz50.head()
```

```python
plt.figure(figsize=(16,8))
plt.scatter(sz50['PCA-5'],sz50['000016.ZICN'],c=mpl_dates)
lin_reg = np.polyval(np.polyfit(sz50['PCA-5'],
                                sz50['000016.ZICN'], 1),
                                sz50['PCA-5'])
plt.plot(sz50['PCA-5'], lin_reg, 'r', lw=3)
plt.grid(True)
plt.xlabel('PCA-5')
plt.ylabel('sz50')
plt.colorbar(ticks=mpl.dates.DayLocator(interval=250),
                format=mpl.dates.DateFormatter('%d %b %y'))
```

上图揭示了2017年初存在某种结构断裂，因此，我们将整个时间轴分为两个子时期，然后分别实施一次早期回归和晚期回归。

```python
cut_date = '2017-1-1'
early_pca = sz50[sz50.index < cut_date]['PCA-5']
early_reg = np.polyval(np.polyfit(early_pca,
                sz50['000016.ZICN'][sz50.index < cut_date], 1),
                early_pca)
```

```python
late_pca = sz50[sz50.index >= cut_date]['PCA-5']
late_reg = np.polyval(np.polyfit(late_pca,
                sz50['000016.ZICN'][sz50.index >= cut_date], 1),
                late_pca)
```

```python
plt.figure(figsize=(16,8))
plt.scatter(sz50['PCA-5'], sz50['000016.ZICN'], c=mpl_dates)
plt.plot(early_pca, early_reg, 'r', lw=3)
plt.plot(late_pca, late_reg, 'r', lw=3)
plt.grid(True)
plt.xlabel('PCA-5')
plt.ylabel('sz50')
plt.colorbar(ticks=mpl.dates.DayLocator(interval=250),
                format=mpl.dates.DateFormatter('%d %b %y'))
```
