```python
#下面是一个简单的数据处理案例，假设我们有一个包含学生信息的列表，需要对数据进行处理：


# 原始数据
students = [
    {'name': 'Alice', 'age': 18, 'gender': 'Female', 'grade': 85},
    {'name': 'Bob', 'age': 17, 'gender': 'Male', 'grade': 92},
    {'name': 'Claire', 'age': 18, 'gender': 'Female', 'grade': 78},
    {'name': 'David', 'age': 17, 'gender': 'Male', 'grade': 88},
    {'name': 'Emma', 'age': 16, 'gender': 'Female', 'grade': 95}
]

# 数据处理
female_students = []
male_students = []
average_grade = 0

for student in students:
    average_grade += student['grade']  # 计算总成绩

    if student['gender'] == 'Female':
        female_students.append(student)  # 分组：女生
    else:
        male_students.append(student)  # 分组：男生

average_grade /= len(students)  # 计算平均成绩

# 打印处理结果
print("Female Students:")
for student in female_students:
    print("Name: {student['name']}, Age: {student['age']}, Grade: {student['grade']}")

print("\nMale Students:")
for student in male_students:
    print("Name: {student['name']}, Age: {student['age']}, Grade: {student['grade']}")

print("\nAverage Grade:", average_grade)


#以上代码根据学生的性别将数据分组，并计算所有学生的平均成绩。最后，分别打印出女生和男生的信息，以及平均成绩。这个例子展示了如何使用基本的数据处理技巧，如遍历、条件判断和累加，对数据进行简单的分组和计算。
```

```python
class StockTradingAccount:
    def __init__(self, balance):
        self.balance = balance
        self.stocks = {}
        self.history = []

    def buy_stock(self, symbol, quantity, price):
        cost = quantity * price
        if cost > self.balance:
            raise ValueError('Insufficient balance')
        if symbol not in self.stocks:
            self.stocks[symbol] = quantity
        else:
            self.stocks[symbol] += quantity
        self.balance -= cost
        self.history.append(('buy', symbol, quantity, price))

    def sell_stock(self, symbol, quantity, price):
        if symbol not in self.stocks or self.stocks[symbol] < quantity:
            raise ValueError('Insufficient stocks')
        revenue = quantity * price
        self.stocks[symbol] -= quantity
        if self.stocks[symbol] == 0:
            del self.stocks[symbol]
        self.balance += revenue
        self.history.append(('sell', symbol, quantity, price))

    def get_balance(self):
        return self.balance

    def get_portfolio(self):
        return self.stocks

    def get_history(self):
        return self.history
# 创建交易账户
account = StockTradingAccount(10000)

# 买入 100 股 AAPL，价格为 150
account.buy_stock('AAPL', 100, 150)

# 卖出 50 股 AAPL，价格为 160
#account.sell_stock('AAPL', 50, 160)

# 买入 200 股 GOOG，价格为 1200
#account.buy_stock('GOOG', 200, 1200)

# 查询账户余额、股票持仓和交易历史
print('Account balance:', account.get_balance())
#print('Stock portfolio:', account.get_portfolio())
#print('Trade history:', account.get_history())


#这个类表示一个银行账户
```
