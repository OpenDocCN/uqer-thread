# 你喜欢股票和债券的混合基金吗？

我不喜欢。因为我觉得，资产的大类配置是我可以干的事情，而择股和择时是基金经理应该干的事情。

但是看似简单的指数基金，以中证500为例，比如天弘中证，500ETF，指数本身公布的数值。

大体上，我们会有观念是：

场外基金可能会在碰到大额的申赎的时候，会巨幅的偏离指数原来的涨跌幅；

500ETF这种，主要风险一是流动性，二是折溢价（但这个场内基金比较大，流动性风险较小）；

指数本身公布的数值，比如沪深300，对于红利分配，它是直接在指数中扣除，所以涨幅会比真实值小。

注：500ETF当年有过一次拆细，我懒，就直接用拆细之后日期的数据，因为已经可以明显的看出差别了。但是，别看谁在上谁在下以定义谁比谁好（天知道起点相同是否合理，如果按照终点相同再来倒推起点的话，就好坏倒转了），主要是看两根线偏离了多远。
