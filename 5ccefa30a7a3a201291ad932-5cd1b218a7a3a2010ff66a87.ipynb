# Momentum与Contrarian策略

此策略来自于UQER的[帮助文档](https://uqer.datayes.com/help/faq/#Momentum策略)
* **Momentum策略思路**

业绩好的股票会继续保持其上涨的势头，业绩差的股票会保持其下跌的势头。

每次调仓将股票按照前一段时间的累计收益率排序并分组，买入历史累计收益最高的那一组

* **Contrarian策略思路**

股票在经过一段时间的上涨之后会出现回落，一段时间的下跌之后会出现反弹。

每次调仓将股票按照前一段时间的累计收益率排序并分组，买入历史累计收益最低的那一组。

```python
import pandas as pd
import time
from CAL.PyCAL import * 

start = '2010-08-01'                     # 回测起始时间
end = '2019-04-30'                       # 回测结束时间
universe = DynamicUniverse('SH50')       # 证券池，支持股票和基金、期货
benchmark = 'SH50'                       # 策略参考基准
freq = 'd'                               # 'd'表示使用日频率回测，'m'表示使用分钟频率回测
refresh_rate = 10                 # 执行handle_data的时间间隔
capital_base=100000
accounts = {
    'stock_account': AccountConfig(account_type='security', capital_base=100000)
}

# 调用quick_backtest方法
# a、回测初始化配置，在代码模式进行策略回测时使用
sim_params = quartz.SimulationParameters(start, end, benchmark, universe, capital_base, refresh_rate=refresh_rate, accounts=accounts)
# b、获取回测行情数据
data = quartz.get_backtest_data(sim_params)
# c、运行结果
results = {}

# d、遍历每一组，进行回测
for quantile_five in range(1, 6):  

    def initialize(context):                 # 初始化策略运行环境
        pass

    def handle_data(context):                # 核心策略逻辑
        account = context.get_account('stock_account')
        current_universe = context.get_universe(exclude_halt=True)
        history = context.history(symbol='all', attribute='closePrice', time_range=20, freq='d', style='ast')['closePrice']

        # 计算累计收益
        factor_data = {'symbol':[], 'c_ret':[]}
        for stock in current_universe:
            if history[stock][0]:
                factor_data['symbol'].append(stock)
                factor_data['c_ret'].append(history[stock][-1]/history[stock][0])

        # 排序，
        factor_data = pd.DataFrame(factor_data).sort(columns='c_ret',ascending = True).reset_index()        
        factor_data.rename(columns={"c_ret":'factor'}, inplace=True) # 列重命名
        factor_data = factor_data.set_index('symbol', drop=True)     # 把'symbol'设置为索引
        q_min = factor_data['factor'].quantile((quantile_five-1)*0.2)
        q_max = factor_data['factor'].quantile(quantile_five*0.2)
        my_univ = factor_data[(factor_data['factor']>=q_min) & (factor_data['factor']<q_max)].index.values#目标证券池
       
        # 交易部分
        #获取所有账户持仓，exclude_halt是否移除持仓中停牌的资产
        current_positions = account.get_positions(exclude_halt=True)
        # 如果持仓股票不在当期目标证券池里面，全部卖出
        sell_list = [stk for stk in current_positions if stk not in my_univ]
        for stk in sell_list:
            account.order_to(stk,0)    
        # 在目标股票池中的，等权买入
        buy_list = [stk for stk in my_univ if stk not in current_positions]
        for stk in buy_list:
            account.order_pct_to(stk, 1.0/len(buy_list))
            
    # 生成策略对象
    strategy = quartz.TradingStrategy(initialize, handle_data)
    
    # 开始回测
    bt, perf = quartz.quick_backtest(sim_params, strategy, data=data)

    # 保存运行结果，1为因子最强组，5为因子最弱组
    results[6 - quantile_five] = {'max_drawdown': perf['max_drawdown'], 'sharpe': perf['sharpe'], 'alpha': perf['alpha'], 'beta': perf['beta'], 'information_ratio': perf['information_ratio'], 'annualized_return': perf['annualized_return'],'benchmark_annualized_return': perf['benchmark_annualized_return'], 'volatility': perf['volatility'],'bt': bt}
    print "\r当前进度: {:.2%}".format(quantile_five/5.0)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
sns.set_style('white')

fig = plt.figure(figsize=(10,8))#图像的大小设置
fig.set_tight_layout(True)#调整图像边缘及图像间的空白间隔，图像外部边缘的调整可以使用plt.tight_layout()进行自动控制
ax1 = fig.add_subplot(211)#第一个参数子图总行数，第二个参数子图总列数，第三个参数子图位置
ax1.grid() # 面向对象的方式

for qt in results:
    bt = results[qt]['bt']

    data = bt[[u'tradeDate',u'portfolio_value',u'benchmark_return']]
    data['portfolio_return'] = data.portfolio_value/data.portfolio_value.shift(1) - 1.0   
    data['portfolio_return'].ix[0] = data['portfolio_value'].ix[0]/	100000.0 - 1.0
    data['portfolio'] = data.portfolio_return + 1.0     
    data['portfolio'] = data.portfolio.cumprod()          # 总头寸不对冲时的净值序列
    data['benchmark'] = data.benchmark_return + 1.0
    data['benchmark'] = data.benchmark.cumprod()          # benchmark的净值序列
    
    ax1.plot(data['tradeDate'], data[['portfolio']], label=str(qt))

    
ax1.legend(loc=0)
ax1.set_ylabel(u"净值", fontproperties=font, fontsize=16)
ax1.set_title(u"因子不同五分位数分组选股净值走势", fontproperties=font, fontsize=16)


# results 转换为 DataFrame
results_pd = pd.DataFrame(results).T.sort_index()
results_pd = results_pd[[u'alpha', u'beta', u'information_ratio', u'sharpe', u'annualized_return', u'benchmark_annualized_return',u'volatility',u'max_drawdown']]
results_pd.columns = [u'Alpha', u'Beta', u'信息比率', u'夏普比率',  u'年化收益',u'基准年化收益', u'策略年化波动率',u'最大回撤']
results_pd.index.name = u'五分位组别'
print results_pd.to_html()
```

一段时间的累计收益率进行五分位分组回测，收益率最大的一组策略看起来效果不错，当使用此指标来用反转策略时效果并不好。另外五分位的分组回测并没有呈现一定的相关性，说明利用累积收益率来选股不是线性关系。但可以说明，买贵的比买便宜的更好。
