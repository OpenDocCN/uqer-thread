# 重复值，替换值，画图

```python
#1.3移除重复值     DataFrame.drop_duplicates(subset=None(表示考虑所有列，列标签可选）, [keep='first'（保留第一次出现的重复项，默认值）,'last'（删除重复项，仅保留最后一次出现的重复项）,'False'(删除所有重复项）], inplace=False)
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
data={'name':['张三','李四','张三','小明'],
      'sex':['female','male','female','male'],
      'year':[2001,2002,2001,2002],
      'city':['北京','上海','北京','北京']}
df1=DataFrame(data)
print(df1)

print(df1.duplicated())#哪一行有重复值，哪一行就显示True，其余显示False

print('df1.drop_duplicates():')
print(df1.drop_duplicates())#删除有重复项的那一行，注意与上一行单词不同

print('df1.drop_duplicates(keep=False):')
print(df1.drop_duplicates(keep=False))

print(df1.drop_duplicates(['sex','year']))#确定具体哪一列相同的情况下进行删除重复项,默认删除后来重复的行
print(df1.drop_duplicates(['sex','year'],keep='last'))#保留后来出现重复的行


data = pd.DataFrame({'A':['a','b','c','c'],'B':[1,1,2,2]})
print(data)
print(data.drop_duplicates(keep='last'))
```

```python
#1.4替换值     replace(oldvalue, newvalue, count)
from pandas import DataFrame
data={'name':['张三','李四','王五','小明'],
      'sex':['female','male','','male'],
      'year':[2001,2002,2001,2002],
      'city':['北京','上海','','北京']}
df1=DataFrame(data)
print('df:')
print(df1)
print(df1.replace('','不详'))#将空格替换成不详

print(df1.replace(['',2001],['不详',2002]))#将空格替换成不详，将2001替换成2002
print(df1.replace({'':'不详',2001:2002}))#将空格替换成不详，将2001替换成2002，两种结果一样
```

```python
#1.5数据转换
data={'name':['张三','李四','王五','小明'],
      'math':[79,52,63,92]
     }
df2=DataFrame(data)
print(df2)

def f(x):
    if x>=90:
        return'优秀'
    elif 70<=x<90:
        return'良好'
    elif 60<=x<70:
        return'及格'
    else:
        return'不合格'

df2['class']=df2['math'].map(f)#根据成绩增加一列
print(df2)
df2['class']=df2['math'].apply(f)
print(df2)
```
