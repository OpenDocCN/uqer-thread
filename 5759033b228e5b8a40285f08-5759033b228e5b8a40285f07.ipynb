```python
import numpy as np
import math
import pandas as pd
#import datetime
#from scipy import interpolate
from matplotlib import pylab
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
pd.set_option('display.width', 800)
class Equation() :
    def __init__(self,D,a,b,t0,t1,Initial_condition,Bondary_conditionL,Bondary_conditionR):
        
        self.D=D
        self.X=b-a
        self.T=t1-t0
        self.i_c=Initial_condition
        self.bcl=Bondary_conditionL
        self.bcr=Bondary_conditionR
    
    
class Explictmethod(Equation):
    def __init__(self,M,N,D,a,b,t0,t1,Initial_condition,Bondary_conditionL,Bondary_conditionR):
        Equation.__init__(self,D,a,b,t0,t1,Initial_condition,Bondary_conditionL,Bondary_conditionR)
        self.M=M
        self.N=N
        
        self.k=self.T/self.M
        self.h=self.X/self.N
        
        self.xarray=np.linspace(a,b,self.N+1)
        self.yarray=np.linspace(t0,t1,self.M+1)
        self.U=np.zeros([M+1,N+1])
        self.U[0,:]=map(self.i_c,self.xarray)
        self.rho=(self.D*self.k)/self.h**2
        for i in range(int(self.M+1)):
            self.U[i][0]=self.bcl(self.U[i][0])
            self.U[i][-1]=self.bcr(self.U[i][-1])
        
    def roll_back(self):
        for t in range(int(self.M)):
            self.U[t+1][1:self.N]=self.rho*self.U[t][0:self.N-1]+(1-2*self.rho)*self.U[t][1:self.N]+self.rho*self.U[t][2:self.N+1]
#            print t
        
        return self.U
        
        
        
        
        
c=Explictmethod(2500.,25.,1.,0.,math.pi,0.,1.,lambda x:math.sin(x),lambda x:0,lambda x:0) 
print c.roll_back()
pylab.plot(c.xarray,c.U[-1][:]) 

x_plot, t_plot = np.meshgrid(c.xarray,c.yarray)
fig= pylab.figure(figsize = (16,10))
ax = fig.add_subplot(1, 1, 1, projection = '3d')
cutoff = int(0.2 / c.k) + 1
surface = ax.plot_surface(x_plot[:,:cutoff], t_plot[:,:cutoff], c.U[:,:cutoff], cmap=cm.jet)
ax.set_xlabel("$x$", fontdict={"size":18})
ax.set_ylabel(r"$\tau$", fontdict={"size":18})
ax.set_zlabel(r"$U$", fontdict={"size":18})
ax.set_title(u"热传导方程 $u_\\tau = u_{xx}$, $\\rho = 0.521$" , fontproperties = font)
fig.colorbar(surface,shrink=0.75)

```
