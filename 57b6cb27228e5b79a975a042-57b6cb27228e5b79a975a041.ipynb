在社区里看到了@alpha_cat大神分享的[国海证券《因子预测能力初探》](https://uqer.datayes.com/community/share/57a3ea90228e5b9b8fa86b0f)，他搭建的分析框架非常好用，我这里在其代码的基础上进行了扩展，新加了观察因子横截面alpha的函数

结合alpha_cat本来的rank_ic，对比着看可以得到更多有关因子有效性的信息。从这个角度发散开去的话，可以逐渐完善单因子效果的分析框架

```python
import matplotlib
import matplotlib.pyplot as plt
import scipy.stats as st
import statsmodels.api as sm
import numpy as np
import pandas as pd
import seaborn

IDXMAP = {
    'SH50': '000016',
    'HS300': '000300',
    'ZZ500': '000905',
}

def get_weekend_trading_day_list(begin_day="20140101", end_day="20160801"):
    """获取每周的最后一个交易日"""
    cal_dates = DataAPI.TradeCalGet(exchangeCD=u"XSHG", beginDate=begin_day, endDate=end_day, field="calendarDate,isWeekEnd")
    trading_days = cal_dates[cal_dates['isWeekEnd']==1]["calendarDate"].tolist()
    trading_days = [day[0:4] + day[5:7] + day[8:] for day in trading_days]  # 更改日期的格式，将'2014-01-03'转化为'20140103'
    return trading_days

def get_index_cons(day="20140110", ticker="000300"):
    """获取当日指数的成份股，结果用字符串拼接起来方便后续的传参"""
    cons_id_df = DataAPI.IdxConsGet(secID=u"",ticker=ticker,intoDate=day,isNew=u"",field=u"consID",pandas="1")
    cons_id_ls = cons_id_df["consID"].tolist()
    return ",".join(cons_id_ls)

def plot_result(result_df, factor_name, ticker):
    """可视化最后的结果"""

    matplotlib.style.use('ggplot')
    result_df["date"] = result_df["date"].apply(str)
    result_df.index = pd.to_datetime(result_df["date"])
    result_df.plot(colors=['royalblue', 'crimson'], figsize=(15, 6)).set_title(factor_name+" in "+ticker)
    plt.axhline(0, color='Gray')
```

```python
def plot_rank_ic(factor_name, ticker="000300", beging_day="20140101", end_day = "20160801"):
    """主函数，计算rank-ic并且画图
       @factor_name 为因子名， 如"RSI","ADX"等
       @ticker 指数代码，如000300代表沪深300指数
       @beging_day 起始时间
       @end_day 结束时间
       """
    
    ticker = IDXMAP.get(ticker, ticker)
    
    trading_days = get_weekend_trading_day_list(beging_day, end_day) 
    ic_ls = [] #保存计算结果，最终转化为DataFrame格式
    factor_api_field = "secID," + factor_name

    for i in xrange(len(trading_days)-1):
        sons_id_str = get_index_cons(trading_days[i],ticker)
        factor_df = DataAPI.MktStockFactorsOneDayGet(tradeDate=trading_days[i],secID=sons_id_str,ticker=u"",field=factor_api_field,pandas="1")  #获取每周最后一个交易日的因子值
        weekly_return = DataAPI.MktEquwAdjGet(secID=sons_id_str,beginDate=trading_days[i+1],endDate=trading_days[i+1],field=u"secID,return",pandas="1") # 获取相应股票未来一周的收益
        result = factor_df.merge(weekly_return,on='secID', how="inner")
        result[result["return"]==0] = None   #如果收益是0，非常有可能该股票一直处于停牌中，这里把种情况剔除
        result.dropna(inplace=True)

        ic, p_value = st.pearsonr(result[factor_name].rank(),result["return"].rank())   #计算Rank-IC
        
        temp_dict = {}
        temp_dict["date"] = trading_days[i]
        temp_dict["Rank-IC"] = ic
        ic_ls.append(temp_dict)
    
    result_df = pd.DataFrame(ic_ls)
    result_df["MovingAverage"] = pd.rolling_mean(result_df["Rank-IC"], window=20,min_periods=10)
    plot_result(result_df, factor_name,ticker)
```

```python
def plot_cs_alpha(factor_name, ticker="000300", beging_day="20140101", end_day = "20160801"):
    """主函数，计算crosssection-alpha并且画图
       @factor_name 为因子名， 如"RSI","ADX"等
       @ticker 指数代码，如000300代表沪深300指数
       @beging_day 起始时间
       @end_day 结束时间
       """
    
    ticker = IDXMAP.get(ticker, ticker)
    
    trading_days = get_weekend_trading_day_list(beging_day, end_day) 
    ic_ls = []  # 保存计算结果，最终转化为DataFrame格式
    factor_api_field = "secID," + factor_name

    for i in xrange(len(trading_days)-1):
        sons_id_str = get_index_cons(trading_days[i],ticker)
        factor_df = DataAPI.MktStockFactorsOneDayGet(tradeDate=trading_days[i],secID=sons_id_str,ticker=u"",field=factor_api_field,pandas="1")  # 获取每周最后一个交易日的因子值
        weekly_return = DataAPI.MktEquwAdjGet(secID=sons_id_str,beginDate=trading_days[i+1],endDate=trading_days[i+1],field=u"secID,return",pandas="1")  # 获取相应股票未来一周的收益
        result = factor_df.merge(weekly_return, on='secID', how="inner")
        result[result["return"]==0] = None   # 如果收益是0，非常有可能该股票一直处于停牌中，这里把种情况剔除
        result.dropna(inplace=True)

        # ic, p_value = st.pearsonr(result[factor_name].rank(),result["return"].rank())   #计算Rank-IC
        
        x = np.column_stack((np.array(result[factor_name].tolist()),))
        x = sm.add_constant(x, prepend=True)
        y = np.array(result['return'].tolist())
        res = sm.OLS(y,x).fit(method = 'qr')
        
        temp_dict = {}
        temp_dict["date"] = trading_days[i]
        temp_dict["cs_alpha"] = res.params[0]
        ic_ls.append(temp_dict)
    
    result_df = pd.DataFrame(ic_ls)
    result_df["MA_cs_alpha"] = pd.rolling_mean(result_df["cs_alpha"], window=20,min_periods=10)
    plot_result(result_df, factor_name,ticker)
```

```python
plot_rank_ic("ADX", beging_day="20150101", end_day = "20160801")  # 沪深300
plot_cs_alpha("ADX", beging_day="20150101", end_day = "20160801")  # 沪深300
```
