# 指数分化情况历史分析

## 指数分化情况历史分析
---
从2016年10月开始的，一两个月的大蓝筹股行情，上证50、上证综指与中证500、创业板的走势产生了较大的分化。经过12月2日开始的大幅下跌调整，昨日(2016年12月15日)，该分化的方向出现了明显的反向，上证50下跌超过2%，而中证500与创业板都有超过0.5%的涨幅。
<br>
我们不禁要想，该分化会继续持续，还是又重新回归蓝筹行情，亦或指数继续大幅下跌？本文旨在分析指数历史分化情况，以期对后市的变化做一个指导。

```python
import numpy as np
import pandas as pd
from datetime import datetime
from dateutil.parser import parse
import matplotlib.pyplot as plt
import seaborn as sn
from CAL.PyCAL import *    # CAL.PyCAL中包含font
cal = Calendar('China.SSE')
font.set_size(15)
sn.set_style('white')
```

```python
def get_index_quotation(index_ticker,start_date,end_date):
    df_index = DataAPI.JY.MktIdxdJYGet(beginTime=u"",endTime=u"",indexID=index_ticker,secIDSrc=u"",
                                       startDate=start_date,endDate=end_date,field=u"tradeDate,closeIndex",pandas="1")
    df_index['tradeDate'] = df_index['tradeDate'].map(lambda x: parse(x))
    df_index = df_index.sort_values(['tradeDate'])
    return df_index
```

```python
def mutil_index_compare(index_list,colors = ['dodgerblue','deeppink','gold','lime'],legends=[u'上证综指',u'沪深300',u'中证500',u'上证50']):
    
    fig = plt.figure(figsize=(14,7))
    ax = fig.add_subplot(111)
    for i in range(len(index_list)):
        ax.plot(index_list[i]['tradeDate'].values,index_list[i]['net_value'].values,colors[i],linewidth=2.5)
    ax.legend(legends,prop=font,loc='upper left')
    ax.set_title(u"指数净值走势", fontproperties=font, fontsize=18)
```

```python
def start_end_index_plot(start_date,end_date):
    index_list = []
    for index in ['000001.ZICN','000300.ZICN','000905.ZICN','000016.ZICN']:
        df_index = get_index_quotation(index,start_date,end_date)
        df_index['net_value'] = df_index['closeIndex'] / df_index['closeIndex'].iloc[0]
        index_list.append(df_index)
    mutil_index_compare(index_list)
```

### 指数分化时序图
---
接下来我们看下指数分化的历史情况，此处我们以上证50与中证500在一个时间窗口内的累计收益的差值作为分化程度的衡量标准，来看历史上指数的分化情况。

```python
def get_index_quotation(index_ticker,start_date,end_date):
    df_index = DataAPI.JY.MktIdxdJYGet(beginTime=u"",endTime=u"",indexID=index_ticker,secIDSrc=u"",
                                       startDate=start_date,endDate=end_date,field=u"tradeDate,closeIndex",pandas="1")
    df_index['tradeDate'] = df_index['tradeDate'].map(lambda x: parse(x))
    df_index = df_index.sort_values(['tradeDate'])
    return df_index
```

```python
df_016 = get_index_quotation('000016.ZICN','20050101','20161215')
df_905 = get_index_quotation('000905.ZICN','20050101','20161215')
```

```python
def rolling_cum_yield(df_016,df_905,window):
    
    rolling_summary = [df_905['tradeDate'],pd.rolling_apply(df_905['closeIndex'],window,lambda x:x[-1] / x[0] -1).rename(u"中证500滚动收益"),pd.rolling_apply(df_016['closeIndex'],window,lambda x:x[-1] / x[0] -1).rename(u"上证50滚动收益")]
    rolling_summary = pd.concat(rolling_summary,axis=1)
    rolling_summary[u"收益差值"] = rolling_summary[u"上证50滚动收益"] - rolling_summary[u"中证500滚动收益"]
    return rolling_summary
```

```python
def index_division_plot(df_016,df_905,window):
    index_rolling_summary = rolling_cum_yield(df_016,df_905,window)
    index_division['tradeDate'] = pd.to_datetime(index_division['tradeDate'])
    fig = plt.figure(figsize=(20,7))
    ax = fig.add_subplot(111)
    ax2=ax.twinx()
    ax.plot(index_rolling_summary.tradeDate,index_rolling_summary[u"收益差值"].values,'dodgerblue',linewidth=2.5)
    ax.plot(index_rolling_summary.tradeDate,[index_rolling_summary[u"收益差值"].values[-1]]*len(index_rolling_summary),'r--')
    ax.plot(index_rolling_summary.tradeDate,[0]*len(index_rolling_summary),'k--')
    ax2.plot(df_905['tradeDate'].values,df_905['closeIndex'].values,'gold',linewidth=2)
    ax.legend([u"累计收益差值",u"当前收益差值"],loc='upper left',bbox_to_anchor=(0.01,0.99),prop=font,frameon= False,handlelength = 3,handletextpad=0.1,borderpad = 0.1)
    ax2.legend([u"中证500(右轴)"],loc='upper left',bbox_to_anchor=(0.15,0.99),prop=font,frameon= False,handlelength = 3,handletextpad=0.1,borderpad = 0.1)
    ax.set_title(str(window) + u"个交易日上证50与中证500累计收益差值走势", fontproperties=font, fontsize=18)
```

```python
index_division_plot(df_016,df_905,15)
```

```python
index_division_plot(df_016,df_905,25)
```

可以看出相比历史较大上的分化行情，此次的分化情况并谈不上极端，并且随着昨日创业板率先反弹，上证50与中证500的累计收益差值已经开始缩减。但总体来看收益差值仍处于历史较高位置，后市有望走出反向分化，今日（2016年12月16）也是创业板带头反弹，个人认为会中小创会有一波行情，即使没有一波行情，上证50与中证500的累计收益差值进一步缩小概率较大。
