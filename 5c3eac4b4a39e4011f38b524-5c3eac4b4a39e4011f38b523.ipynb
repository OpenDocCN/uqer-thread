# 银行结构化理财产品分析——以交通银行原油1M鲨鱼鳍结构化产品为例
## 内容简介
- 以交通银行原油1M鲨鱼鳍结构化产品为例，用蒙特卡洛模拟的方法对看涨和看跌两种产品理论价格进行计算
	- 结果显示看跌和看涨两种产品的溢价率分别为0.02%和0.01%
	- 且两种产品的平均年化收益率分别为3.22%和3.25%
	- 认为产品对于投资者可在风险可控范围内追求更高的投资收益，而对发行方也可在可控风险范围内获得稳定收益

## 小结
- 根据结果发现在实际中，看到的类似此类的理财产品实际收益可能并没有“想象”中这么高
- 虽然产品给出的最低收益率为2.50%，但实际在扣除托管费用（月0.05%）、销售手续费（年化0.5%）后，实际能保证的年化收益率也就1.4%
- 注意不要同时买看涨和看跌两种产品，以为可以拿到一个最低收益，另一个拿到最高的收益，假如产品发生敲出，那么两个综合的收益其实很低
	- 结果中显示同时买两种产品能拿到的最高收益率为5.78%，而最低收益率为1.40%，平均年化收益为3.23%

- 产品挂钩上交所原油期货1903合约
- 若投资的资产按时收回全部资金，产品收益率计算如下：
    - 若单期产品观察期内挂钩标的收盘价与当期产品投资起始日挂钩标的收盘价的比值曾小于障碍价，则单期产品预期年化收益率等于预期敲出收益率；
    - 若单期产品观察期内挂钩标的收盘价与当期产品投资起始日挂钩标的收盘价的比值从未小于障碍价，则单期产品预期年化收益率计算规则如下：
        - 若单期产品观察日挂钩标的收盘价与当期产品投资起始日挂钩标的收盘价的比值大于行权价，单期产品预期年化收益率等于预期最低收益率（前述计算所得预期年化收益率保留两位小数，舍去第三位及之后小数）；
        - 若单期产品观察日挂钩标的收盘价与当期产品投资起始日挂钩标的收盘价的比值小于或等于行权价，单期产品预期年化收益率等于预期最低收益率+参与率*（行权价-当期产品观察日挂钩标的收盘价/当期产品投资起始日挂钩标的收盘价）（前述计算所得预期年化收益率保留两位小数，舍去第三位及之后小数）

- 对于上述结构化产品，类似于一个固定收益部分和一个浮动收益（敲出期权衍生品）的组合
- 下面以看跌鲨鱼鳍为例，举例说明其收益计算：
    - 假定某投资者认购单期产品 1,000,000 元人民币，单期产品投资起始日为 2019年 1 月 17 日，单期产品投资到期日为 2019 年 2 月 20 日，单期产品预期投资期限为34 天，单期产品观察日为 2019 年 2 月 15 日，单期产品观察期为 2019 年 1 月 17 日（含）到 2019 年 2 月 15 日（含）之间的每一个工作日。假设单期产品投资起始日挂钩标的参考价格为 400，可能的投资情况如下：
        - 情景一：若单期产品观察期内的原油1903合约收盘价曾小于376(400 * 94%=376)，则单期产品预期年化收益率为5.00%，投资者收回产品本金，并获得产品收益人民币4,657.53元（1,000,000 * 5.00% * 34/365＝4,657.53）；
        - 情景二：若单期产品观察期内的原油1903合约收盘价从未小于376（400 * 94.00%=376），且在观察日当天原油1903合约的收盘价为404，高于400（400 * 100.00%=400），则单期产品预期年化收益率为2.50%，投资者收回产品本金，并获得产品收益人民币2,328.77元（1,000,000 * 2.50% * 34/365＝2,328.77）；
        - 情景三：若单期产品观察期内的原油1903合约收盘价从未小于376（400 * 94.00%=376），且在观察日当天原油1903合约的收盘价为384，低于400（ 400 * 100.00%=400 ） ， 则 单 期 产 品 预 期 年 化 收 益 率 为 6.50%（2.50%+100% * (100.00%-384/400)=6.50%），投资者收回产品本金，并获得产品收益人民币6,054.79元（1,000,000 * 6.50% * 34/365＝6,054.79）；

- 对于产品的定价，首先这里进行一些说明，由于产品的特性，难以分别对固收部分和衍生品部分单独定价，故此这里将二者视为整体，通过蒙特卡洛模拟的方法对标的资产（原油期货1903合约）收盘价进行模拟，根据结构化产品的特点来计算期末产品的收益，得到期末现金收益后进行贴现，得到产品的定价。
    - 产品收益要减去发行方收取的费用；
    - 产品存在收益上限，即并不能像看涨期权在理论上可以获得无限大的收益

- 要使用蒙特卡洛模拟方法对产品进行定价，首先要确定利用何种随机过程对标的资产的价格进行模拟，且需要对随机过程模型的参数进行确定，这里根据数据特征（原油1903合约收盘价）选取几何布朗运动进行模拟，利用历史数据对这些参数进行确定。选取了2018年3月26日-2019年1月9日的历史数据进行参数拟合，这里选择到2019年1月9日的原因在于产品的认购期为2019年1月10日起。
- 从描述性统计结果中可以得到几何布朗运动的波动率为0.294821，且对于蒙特卡洛模拟过程，这里假设模拟环境为风险中性的，且此处的无风险利率采用2019年1月9日的1年期国债收益率，为2.38%。
- 对于此产品，其在产品成立日之前，已经开始预售，预售期为2019年1月10日-2019年1月16日，在预售期间，标的资产也在交易，而产品收益率推出后收益率区间已经限定，从而在发行方的角度来讲，产品的定价最晚要根据1月9日原油期货1903合约收盘价来确定。所以在标的资产路径模拟的过程中需要以2019年1月9日的收盘价作为初始价格来进行模拟，步数为22天（截至观察日2019年1月15日共22个交易日，此处去除了春节休市日期，但要注意的是在产品价格贴现计算时的时长为34天，因为产品的收益率取决于观察期和观察日的标的价格），进行100000次路径模拟

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math 
```

```python
# 参数设置
S0 = 416.3
purchase_money = 10000       # 购买金额设置
r = 3.401/100                # 无风险利率，这里采用2019年1月9日的1年期shibor利率
sigma = 0.294821             # 波动率
# 交易参数
paths = 100000
steps_year = 252    # 每年的交易日天数
dt = 1/steps_year
steps = 22          # 模拟步长
discount_t = 34/365  # 贴现步长
# 产品参数设定
participate_call = 90/100   # 看涨产品参与率
participate_put = 100/100   # 看跌产品参与率
knock_rate_call = 107/100   # 看涨产品敲出价格
knock_rate_put = 94/100     # 看跌产品敲出价格
basic_return = 2.5/100      # 最低收益率
knock_return_call = 5.2/100   # 看涨产品敲出收益率
knock_return_put = 5/100      # 看跌产品敲出收益率
highest_return_call = 8.8/100 # 看涨产品最高收益率
highest_return_put = 8.5/100  # 看跌产品最高收益率
host_fee_rate = 0.05/100    # 托管费率(月度)
sale_fee_rate = 0.5/100     # 销售手续费率
```

```python
# 对蒙特卡洛模拟及产品理论价格进行定义

# 函数中paths_save：0表示不保存路径，1表示保存路径，默认为 0; S_start:标的资产初始价格，默认为 S0；risk_neutral：判定是否为风险中性
def monte_carlo_geo_brown(steps_simulate, paths_simulate, paths_save=0, S_start=S0, risk_neutral=True):
    # 是否风险中性判断
    if risk_neutral == True:
        mu_sim = r
    else:
        mu_sim = mu
    # 是否保存模拟路径判断
    if paths_save == 1:
        S_simulate = np.zeros((steps_simulate+1, paths_simulate))
        S_simulate[0] = S_start
        for ii in range(steps_simulate):
            winc = math.sqrt(dt) * np.random.randn(1, paths_simulate)
            S_simulate[ii+1] = S_simulate[ii] + S_simulate[ii] * mu_sim * dt\
            + S_simulate[ii] * winc[0]
    if paths_save == 0:
        S_simulate = np.zeros((1, paths_simulate))
        S_simulate[0] = S_start
        for ii in range(steps_simulate):
            winc = math.sqrt(dt) * np.random.randn(1, paths_simulate)
            S_simulate[0] = S_simulate[0] + S_simulate[0] * mu_sim * dt + S_simulate[0] * winc

    return S_simulate


def cal_value_for_shark(route_sim, S_0=S0, option_kind='call'):
    # 期权价值计算，此处价值先以收益率形式进行表示，最后转换为实际形式
    bool_ = route_sim.copy()
    # 首先判断看涨产品还是看跌产品
    option_kind = option_kind.lower()
    # 看涨情况
    if option_kind == 'call':
        # 设置最低收益
        value = np.ones((1,paths)) * basic_return
        # 筛选敲出部分
        bool_[bool_ > knock_rate_call * route_sim[0]] = np.nan
        bool_ = pd.DataFrame(bool_)
        bool_judge = bool_.isnull().any()
        for ii in range(len(bool_judge)):
            if bool_judge[ii] == True:
                value[0,ii] = knock_return_call
            elif route_sim[-1,ii] > route_sim[0,ii]:
                # 设置最高收益
                value_ = basic_return + participate_call * (route_sim[-1,ii]/route_sim[0,ii] - 1)
                value[0,ii] = min(value_, 0.088)
    # 看跌情况
    elif option_kind == 'put':
        # 设置最低收益
        value = np.ones((1,paths)) * basic_return
        bool_[bool_ < knock_rate_put * route_sim[0]] = np.nan
        bool_ = pd.DataFrame(bool_)
        bool_judge = bool_.isnull().any()
        for jj in range(len(bool_judge)):
            if bool_judge[jj] == True:
                value[0,jj] = knock_return_put
            elif route_sim[-1,jj] < route_sim[0,jj]:
                value_ = basic_return + participate_put * (route_sim[0,jj]/route_sim[-1,jj] - 1)
                value[0,jj] = min(value_, 0.085)
    else:
        print('option_kind enter error!')
    
    # 扣除费率后的收益率
    return_money = purchase_money * (1 - host_fee_rate*(34/30) - sale_fee_rate * (34/365)) + value * purchase_money * discount_t
    # 将收益率形式转换为实际现金流    
    price_ = return_money/(1 + r * discount_t)
    price = np.mean(price_)
    
    return value, return_money, price
```

- 得到的理论定价中显示，看跌和看涨两种产品的溢价率分别为0.02%和0.01%，且两种产品的平均年化收益率分别为3.22%和3.25%
- 且需要注意的在于，在严格意义上此产品的溢价率要更高

    - 对于产品到期后，本金及收益并不能立即到账，也会加大产品的溢价率
    - 在产品说明书的表述中， 发行方有权单方面的对产品的费用率进行调整，且可以单方面的提前终止产品，也就是说，若发行方预计产品会产生极高收益率而导致发行方亏损的情况下，发行方可以单方面终止产品，这样发行方的风险就是稳定可控的。
