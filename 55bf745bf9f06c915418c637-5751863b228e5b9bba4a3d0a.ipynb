```python
团队名称：统计三剑客
策略名称：喵
同学们好。
我们找了几个量价因子，然后尝试用决策树模型来选股。
#Portfolio是Library的名字
发现实盘和样本内有一些差距，下面的比赛要注意过拟合问题。
谢谢大家指教。
```

```python
#coding=utf-8

from numpy import *
from sklearn import tree

def dt(universe):
    X=[]
    y=[]
    for name in universe:
        Data=DataAPI.MktEqudGet(secID=name,ticker=u"",beginDate=u"20121201",endDate=u"20150530",field=u"",pandas="1")
        for i in xrange(len(Data['tradeDate'])-6):
            add=[Data['openPrice'][i]/float(Data['openPrice'][i]),Data['closePrice'][i]/float(Data['openPrice'][i]),Data['highestPrice'][i]/float(Data['openPrice'][i]),Data['lowestPrice'][i]/float(Data['openPrice'][i]),Data['turnoverVol'][i],float(Data['openPrice'][i+6])/Data['openPrice'][i+1]-1]
        
            if sum(isnan(add))==0 and add[-1]!=inf:
                X.append(add[:-1])
                y.append(add[-1])
    clf = tree.DecisionTreeRegressor()
    clf.fit(X,y)
    
    print clf.score(X,y)
    print 'done'
    return clf

#c=dt(universe)
```
