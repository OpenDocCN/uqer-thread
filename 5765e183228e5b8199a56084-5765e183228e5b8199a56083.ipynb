```python
import re
import pandas as pd
import numpy as np
Futu = DataAPI.FutuGet(exchangeCD=u"",secID=u"",ticker=u"",contractStatus="L",contractObject=u"",field=u"",pandas="1")
tikers = Futu['ticker']
p=re.compile('[a-zA-Z]+')
cons = []
tickers=tikers.tolist()
for ticker in tickers:
    a=p.match(ticker)
    if len(cons)==0 or cons[-1]!=a.group():
        cons.append(a.group())
        
n=0
for con in cons:
    sids = DataAPI.FutuGet(exchangeCD=u"",secID=u"",ticker=u"",contractStatus="L",contractObject=con,field=u"",pandas="1")['secID']
    data = {}
    for sid in sids :
        one = DataAPI.MktFutdGet(secID=sid,beginDate='20160401',endDate='20160610',field= ['closePrice','mainCon'])
        if not one['mainCon'].empty and one['mainCon'][len(one['mainCon'])-1]:
            mainCon = sid 
        data.update({sid : one['closePrice']})
    p = pd.DataFrame(data) 
    corr = p.corr()
    if mainCon != '':
        corrs = corr[mainCon]
        for (sid,cor) in zip(corr.columns.tolist(),corrs.tolist()):
            if 1 > cor > 0.99:
                p1 = np.polyfit(data[mainCon],data[sid] , 1)
                if abs(p1[0]-1) < 0.05:
                    n += 1
                    factor1 = np.poly1d(p1)
                    print 'Index : '+str(n)
                    print ' : '.join([mainCon,sid])               
                    print ' : '.join(['cor',str(cor),str(factor1)])
                    print  ' '                    
    mainCon=''
```
