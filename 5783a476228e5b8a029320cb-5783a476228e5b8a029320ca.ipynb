```python
###timedata pretreat
def timefunc(x = 30):
    cal = Calendar('China.SSE')
    dtime = []
    for i in range(x):
        a = '-'+str(i+1)+'B'
        dtime.append(a)
    nowtime = datetime.date.today()
    dctime = []
    for j in dtime:
        dti = cal.advanceDate(nowtime,j ,BizDayConvention.Following).strftime('%Y%m%d')
        dctime.append(dti)   
    return dctime
```

```python
###boll function without st
def bollfunc(x = dtt2):
    a = DataAPI.SecTypeRelGet(typeID=u"101001001001",secID=u"",ticker=u"",field=u"",pandas="1")
    b = DataAPI.MktEqudGet(tradeDate=x,secID=u"",ticker=u"",beginDate=u"",endDate=u"",isOpen="1",field=u"",pandas="1")
    c = DataAPI.MktStockFactorsOneDayGet(tradeDate=x,secID=u"",ticker=u"",field=u"ticker,BollDown,BollUp",pandas="1")
    d = pd.merge(a,b,on =['ticker','secID','exchangeCD','secShortName'])
    e = pd.merge(c,d,on ='ticker')
    e = e.dropna()[e.openPrice != 0]
    f = [u'secShortName', u'ticker',u'tradeDate', u'BollDown', u'BollUp',u'secID',u'openPrice', u'highestPrice', u'lowestPrice', u'closePrice']
    g = e[f]
    g['BollMa'] = (g['BollDown']+ g['BollUp'])/2
    g['Increase'] =100*(g['closePrice'] - g['openPrice'])/g['openPrice']
    g1 = g[(g.lowestPrice< g.BollDown) & (g.Increase) > 0]
    g2 = g[(g.lowestPrice<(0.25*g.BollUp+0.75*g.BollDown)) & (g.BollMa < 5*(g.BollUp - g.BollDown))]
    gg = pd.concat([g1,g2],axis = 0)
    gg = gg[gg.duplicated() == False]
    choose = gg[gg.secShortName.str.contains('S') == False]
    return choose
```

这个是我选股策略的核心。意思就是，股票运行到boll带下方四分之一的地方，而且boll带能承受至少两个涨停板的宽度。意思就是，这是一只目前处于低位的股票，而且，这只票还不错，不会跟一些动起来跟蛆一样的票一样。然后，当票突破了boll下四分之一的地方，就看作股票有了上涨趋势。选出来。

```python
###choose ticker
def choose(x = dtt2):
    a1 = bollfunc(x)
    y = cal.advanceDate(x,'-1B',BizDayConvention.Following).strftime('%Y%m%d')
    a2 = bollfunc(y)
    a = []
    for i in list(a2.ticker):
        if i in list(a1.ticker):
            k = False
            a.append(k)
        else:
            k = True
            a.append(k)
    return a2[a]       
        
```

这行代码意思就是前一天还在boll下方运动，今天突破啦！！！确定突破！进行买入！

```python
###reback chang
def change(x = dtt1):
    y = cal.advanceDate(x,'-1B',BizDayConvention.Following).strftime('%Y%m%d')
    cr= choose(y)
    tickerlist = list(cr.ticker)
    a = [u'secShortName', u'ticker',u'tradeDate', u'secID',u'openPrice', u'highestPrice', u'lowestPrice', u'closePrice']
    b = DataAPI.MktEqudGet(tradeDate=x,secID=u"",ticker=tickerlist,beginDate=u"",endDate=u"",isOpen="1",field= a,pandas="1")
    b['Increase'] =100*(b['closePrice'] - b['openPrice'])/b['openPrice']
    return b.Increase.mean()
    
```

计算选出来的股票第二天的涨幅，作为我的获利空间。

```python
###research
changelist = []
t = 100
timelist = timefunc(t)[::-1]
for i in timelist:
    a = change(i)
    changelist.append(a)
```

```python
p = np.array(changelist)
p = np.nan_to_num(p)
rate = 1
for i in p:
    rate = rate*(1+i/100)
print rate*100-100
```
