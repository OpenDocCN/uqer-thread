# `基础策略`
---
* #### __直观想法__：在各行业中选取表现好的股票，形成资产组合，再利用市场指数对冲，使回报稳定，坐收$\alpha$. 

* #### __实现步骤__：
    * #### 将全部A股按照申万一级分类分为30多个行业，在每个调仓日，在每个行业中选取最好的一只股票持有；
    * #### `行业最佳`的选取可以按照基本面信息，比如本篇的基础策略，就以PE最小作为最好股票的标志。

* #### __策略表现__：
    * #### 可以看见，从2013年以来，策略一直保持相对稳定的超额收益。如果用期指对冲，也能以较快速度离开股灾，甚至在2016年来的震荡行情中能实现10%以上的年化收益；
    * #### 另一方面，策略 $\beta$ 值接近1，这与我们构建纯$\alpha$策略的想法不谋而合。

# `扩展范围`
---
* #### 看到上面PE的表现，我忍不住想看看是不是有其他的基本面因子可以实现更好的$\alpha$收益；
* #### 借助quartz的快速回测，可以继续测试以下因子的表现：
> {'PE': 0,'PB': 0,'PS': 0, 'ROA': 1, 'ROE': 1, 'FEARNG': 1, 'TotalAssetGrowRate': 1, 'TotalProfitGrowRate': 1, 'SalesCostRatio': 0,   
> 'OperatingProfitRatio': 1, 'OperatingProfitGrowRate': 1, 'OperatingExpenseRate': 0, 'NetProfitRatio': 1, 'NetProfitGrowRate': 1}

* #### 上面的dict中，因子代表的值为0表示选取行业中该因子最小的一直股票；而因子值为1表示选取该因子值最大的股票。这也不难理解，PE越小表明股票相对越有价值，而ROA越大才表明公司的盈利能力越好。

```python
#*************************
# libs
#*************************
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
sns.set_style('white')

from CAL.PyCAL import *
titleFont = font.copy()
titleFont.set_size(18)
legendFont = font.copy()
legendFont.set_size(12)

import quartz as qz
from __future__ import division
DataAPI.settings.cache_enabled = True


#*************************
# global vars 
#*************************
__cal = Calendar('China.SSE')

__inds = [IndSW.CaiJueL1, IndSW.ChuanMeiL1, IndSW.DianQiSheBeiL1, IndSW.DianZiL1, IndSW.FangDiChanL1, IndSW.FangZhiFuZhuangL1,
IndSW.FeiYinJinRongL1, IndSW.GangTieL1, IndSW.GongYongShiYeL1, IndSW.GuoFangJunGongL1, IndSW.HuaGongL1, IndSW.JiSuanJiL1,
IndSW.JiXieSheBeiL1, IndSW.JiaYongDianQiL1, IndSW.JianZhuCaiLiaoL1, IndSW.JianZhuJianCaiL1, IndSW.JianZhuZhuangShiL1,
IndSW.JiaoTongYunShuL1, IndSW.JiaoYunSheBeiL1, IndSW.JinRongFuWuL1, IndSW.NongLinMuYuL1, IndSW.QiCheL1, IndSW.QingGongZhiZaoL1, 
IndSW.ShangYeMaoYiL1, IndSW.ShiPinYinLiaoL1, IndSW.TongXinL1, IndSW.XinXiFuWuL1, IndSW.XinXiSheBeiL1, IndSW.XiuXianFuWuL1,
IndSW.YiYaoShengWuL1, IndSW.YinHangL1, IndSW.YouSeJinShuL1, IndSW.ZongHeL1]


__factorList = {'PE': 0,'PB': 0,'PS': 0, 'ROA': 1, 'ROE': 1, 'FEARNG': 1, 'TotalAssetGrowRate': 1, 'TotalProfitGrowRate': 1, 
             'SalesCostRatio': 0, 'OperatingProfitRatio': 1, 'OperatingProfitGrowRate': 1, 'OperatingExpenseRate': 0,
             'NetProfitRatio': 1, 'NetProfitGrowRate': 1}
__currentFactor = 'ROA'


#*************************
# quartz
#*************************
def factorFilter(date):
    factor = __currentFactor
    big = __factorList[factor]
    tmpList = []
    for index in range(len(__inds)):
        try:
            df = DataAPI.MktStockFactorsOneDayGet(date, 
                secID = set_universe(__inds[index], date), field = 'secID,' + factor).sort(factor)
            if big:
                tmpList += [df[df[factor] > 0][-1:].secID.values[0]]
            else:
                tmpList += [df[df[factor] > 0][:1].secID.values[0]]
        except Exception,e:
            # print e," —— Industry %i is empty ——"%__inds[index].ind_code,date
            pass
    return tmpList

    
def initialize(account):                   # 初始化虚拟账户状态
    pass

def handle_data(account):                  # 每个交易日的买入卖出指令
    holdList =  factorFilter(account.previous_date)
    for stk in account.valid_secpos:
        if not stk in holdList:
            order_to(stk, 0)
    buyList = [_ for _ in holdList if 
               (_ in account.universe and _ in account.referencePrice and account.referencePrice[_] > 0)]
    buyList = [_ for _ in buyList if not 'ST' in _]

    for stk in buyList:
        order_to(stk, account.referencePortfolioValue/len(buyList)/account.referencePrice[stk])
    return

start = '2016-01-01'                       # 回测起始时间
end = '2016-08-08'                         # 回测结束时间
benchmark = 'HS300'                        # 策略参考标准
universe = set_universe('A', start)  # 证券池，支持股票和基金
capital_base = 1000000                      # 起始资金
freq = 'd'                                 # 策略类型，'d'表示日间策略使用日线回测，'m'表示日内策略使用分钟线回测
refresh_rate = 20                           # 调仓频率，表示执行handle_data的时间间隔，若freq = 'd'时间间隔的单位为交易日，若freq = 'm'时间间隔为分钟


def factorBackTest(factor):
    global __currentFactor
    __currentFactor = factor
    
    sim_params = qz.sim_condition.env.SimulationParameters(start, end, benchmark, universe, capital_base)
    idxmap, data = qz.sim_condition.data_generator.get_daily_data(sim_params)
    strategy = qz.sim_condition.strategy.TradingStrategy(initialize, handle_data)
    bt, account = qz.quick_backtest(sim_params, strategy, idxmap, data, refresh_rate = refresh_rate)
    perf = qz.perf_parse(bt, account)
    return perf


#*************************
# report
#*************************
res = {}
alpha = {}
for fac in __factorList:
    res[fac] = factorBackTest(fac)
    alpha[fac] = res[fac]['alpha']
    

#************************
# reasoning - 策略归因
#************************
def reasonTimeSplit(beginDate, endDate, period):
    timeSet = []
    tmpStart = beginDate
    tmpEnd = __cal.advanceDate(tmpStart, '+'+str(period)+'B').strftime('%Y%m%d')
    while endDate > tmpEnd:
        timeSet += [(tmpStart, tmpEnd)]
        tmpStart = __cal.advanceDate(tmpEnd, '+'+'1'+'B').strftime('%Y%m%d')
        tmpEnd = __cal.advanceDate(tmpStart, '+'+str(period)+'B').strftime('%Y%m%d')
    timeSet += [(tmpStart, endDate)]
    return timeSet

def reasonBacktest(factor, beginDate, endDate):
    global start
    global end
    start,end = beginDate,endDate
    return factorBackTest(factor)

timeset = reasonTimeSplit('20160105', '20160808', 20)
reasonResDict = {}
for fac in __factorList:
    reasonResDict[fac] = {x[1]: reasonBacktest(fac, x[0], x[1]) for x in timeset}
```

```python
# results wanted
reasonAlpha = pd.DataFrame(columns = reasonResDict.keys(), index = reasonResDict[reasonResDict.keys()[0]].keys())
reasonReturn = pd.DataFrame(columns = reasonResDict.keys(), index = reasonResDict[reasonResDict.keys()[0]].keys())
reasonWinRatio = pd.DataFrame(columns = reasonResDict.keys(), index = reasonResDict[reasonResDict.keys()[0]].keys())

# read from reasonResDict
for x in reasonReturn.columns:
    for y in reasonReturn.index:
        reasonAlpha[x][y] = reasonResDict[x][y]['alpha']
        reasonReturn[x][y] = (reasonResDict[x][y]['cumulative_returns'] - 
                              reasonResDict[x][y]['benchmark_cumulative_returns']).values[-1:][0]
        tmpSeries = reasonResDict[x][y]['returns'] - reasonResDict[x][y]['benchmark_returns']
        reasonWinRatio[x][y] = len(tmpSeries[tmpSeries>0])/len(tmpSeries)
```

* #### 将回测期间设为`2016/01/05 - 2016/08/08`
* #### 下图展示了各个因子的$\alpha$值，可以看见，总体表现不错：投资者关注较多的PE、PB、ROA、ROE等基本面因子都能充当比较好的筛选指标，给出10%以上的$\alpha$. 而成长因子则不那么可靠，这可能给我们一个投资的提示：公司增长快可能并不代表股价增长快，为了获得超额收益，最好持有基本面已经很好的公司股票。

```python
alphaSeries = pd.Series(alpha)
alphaSeries.sort(ascending= False, inplace = True)

def alphaPlot(series = alphaSeries):
    fig = plt.figure(figsize = (1.5 * len(series), 6))
    ax = fig.add_subplot(111)
    xscale = np.arange(0,1,1.0/(len(series)-1)).tolist()+[1]
    barWidth = 0.05

    for idx in range(len(series)):
        if series[series.index[idx]] > 0:
            ax.bar([-barWidth/2.0, xscale[idx] - barWidth/2.0],[0, 100*series[series.index[idx]]],width = barWidth, color = '#F08080')
        else:
            ax.bar([-barWidth/2.0, xscale[idx] - barWidth/2.0],[0, 100*series[series.index[idx]]],width = barWidth, color = 'g')

    ax.set_xticks(xscale)        
    ax.set_xticklabels(series.index)
    # series[series > 0].plot(kind = 'bar', ax = ax, color = 'red', alpha = 0.4)
    # series[series < 0].plot(kind = 'bar', ax = ax, color = 'g')

    fig.autofmt_xdate(rotation = 30, ha = 'right')
    plt.title(u'各因子alpha表现(%)', fontproperties = titleFont, verticalalignment = 'bottom',horizontalalignment = 'center')
    plt.grid(axis = 'y', color = 'grey', alpha = 0.4)
    plt.xlim((-barWidth, 1+barWidth))
    ax.spines['left'].set_color('grey')
    ax.spines['right'].set_color('grey')
    ax.spines['top'].set_color('grey')
    ax.spines['bottom'].set_color('grey')

alphaPlot()
```

* #### 下图展示了各个因子的胜率，即 —— 回测期间正超额收益天数 / 样本总天数。大部分因子胜率在50%以上，胜率较大的因子基本上$\alpha$也会比较高。
* #### PE因子的胜率将近63%，应该是比较让人满意的。

```python
winRatio = pd.Series()
for fac in res:
    perf = res[fac]
    resReport = pd.DataFrame()
    resReport['cumulative_returns'] = perf['cumulative_returns']
    resReport['benchmark_cumulative_returns'] = perf['benchmark_cumulative_returns']
    resReport['ex_cumulative_returns'] = perf['cumulative_returns'] - perf['benchmark_cumulative_returns']
    resReport['ex_returns'] = perf['returns'] - perf['benchmark_returns']
    winRatio[fac] = len(resReport[resReport.ex_returns > 0]) / len(resReport)
    # fig = plt.figure(figsize = (18, 14*0.618))
    # ax = fig.add_subplot(111)
    # resReport.plot(y = ['cumulative_returns', 'benchmark_cumulative_returns'],ax = ax)
    # resReport.plot(y = 'ex_returns', ax = ax)
    # resReport.plot(y= 'ex_cumulative_returns', ax = ax, label = fac + u'因子累计超额收益')
    # plt.xlabel('')
    # plt.legend(prop = legendFont, loc = 'best')
    # plt.grid(axis = 'y', color = 'grey', alpha = 0.4)
    # plt.title(fac + u'表现', fontproperties = titleFont)
winRatio.sort(ascending= False, inplace = True)

def winRatioPlot(series = winRatio):
    fig = plt.figure(figsize = (1.5 * len(series), 6))
    ax = fig.add_subplot(111)
    xscale = np.arange(0,1,1.0/(len(series)-1)).tolist()+[1]
    barWidth = 0.05

    for idx in range(len(series)):
        if series[series.index[idx]] > 0.5:
            ax.bar([-barWidth/2.0, xscale[idx] - barWidth/2.0],[0, 100*series[series.index[idx]] - 50],
                   width = barWidth, color = '#F08080', bottom = 50)
        else:
            ax.bar([-barWidth/2.0, xscale[idx] - barWidth/2.0],[0, 100*series[series.index[idx]] - 50],
                   width = barWidth, color = 'g', bottom = 50)
    ax.set_xticks(xscale)        
    ax.set_xticklabels(series.index)
    # series[series > 0].plot(kind = 'bar', ax = ax, color = 'red', alpha = 0.4)
    # series[series < 0].plot(kind = 'bar', ax = ax, color = 'g')

    fig.autofmt_xdate(rotation = 30, ha = 'right')
    plt.title(u'各因子胜率 —— 相对基准指数超额收益大于0概率(%)', fontproperties = titleFont, verticalalignment = 'bottom',horizontalalignment = 'center')
    plt.grid(axis = 'y', color = 'grey', alpha = 0.4)
    plt.xlim((-barWidth, 1+barWidth))
    ax.spines['left'].set_color('grey')
    ax.spines['right'].set_color('grey')
    ax.spines['top'].set_color('grey')
    ax.spines['bottom'].set_color('grey')
    
winRatioPlot()
```

# `策略稳定性检验`

* #### PE因子在期间每个月的胜率都超过55%，甚至在其他因子表现都不好的最近一个月，其胜率也达到65%，稳定性很不错。

```python
fig = plt.figure(figsize = (12, 12*0.618))
sns.heatmap(reasonWinRatio.astype(float).sort_index().T * 100, cmap= plt.cm.Greys, annot= True, vmax = 80, vmin = 40)
fig.autofmt_xdate()
plt.title(u'各个月份因子胜率(%)', fontproperties = titleFont, verticalalignment = 'bottom')
```

* #### 期间基本每个月PE因子都能有正的超额收益，除4月8日 - 5月10日这一段期间，而这一段期间所有其它因子也没能取得超额收益。这也从另一个角度验证了PE因子表现的稳定性。

```python
fig = plt.figure(figsize = (12, 12*0.618))
sns.heatmap(reasonReturn.astype(float).sort_index().T * 100, cmap= plt.cm.Greys, annot= True)
fig.autofmt_xdate()
plt.title(u'各个月份因子超额收益(%)', fontproperties = titleFont, verticalalignment = 'bottom')
```
