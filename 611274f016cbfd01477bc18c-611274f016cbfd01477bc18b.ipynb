# 基金因子测试框架

# 前言

首先感谢通联数据提供了大量的数据和各种指标用于量化工作，是我见过各种平台中指标最多的，在策略测试中提供了丰富的因子。
对于策略测试来说，有一个重要的步骤是检验因子的单调性，即因子按从大到小排列后，是不是最大的就是收益最好的，因子越小是不是收益就越低。策略模拟本身不能提供这一功能，因此我试图采用研究中模拟的方法取得因子单调性测试结果。

# 测试说明
为满足程序的可塑性，我采用python的类结构。父类提供通用函数和统一接口，子类实现部分特殊函数。
## 父类assetAnalyse
### 输入参数：
- assetsPoolType:资产池类型是动态dynamic还是静态static。
- assetsPool:资产池。如果资产池类型是动态，此处传入指数secID,参数类型是str。如果是静态，传入资产secID的list
- period：测试区间，为元组类型，形如('20180101','20190101')
- benchmark:基准指数
- factor：测试因子。dict类型，key是通联定义的参数名称，如基金是在DataAPI.FundPerfIndicGet函数中的因子名称。value是boolean类型，True代表越小越好，False代表越大越好。
- groupCount:分组数
- window：因子时间窗口。这个参数有点纠结，因为只有基金因子有这个参数，股票应该是没有的，扩展到股票测试时这个参数就没用了。
### 输出结果
输出结果包含每组收益（以1、2、3、4.。表示），benchmark表示基准，mean表示平均买入资产池中资产的收益
- portfolio:类似策略测试结果的总收益，可以导出到excel中绘制曲线。
- rtnPerMonth:每月收益率
- rtnPerYear:每年收益率
- extraRtnPerMonth：每月超额收益
- extraRtnPerYear：每年超额搜狐一
- summary:容易计算的指标，包括总收益和超额收益。
- extraSummary:计算复杂一点的指标，包括年化收益，夏普比率，最大回撤等。
子类目前实现了基金测试子类fundAnalyse，股票测试子类正在研究中。

```python
import pandas as pd
from datetime import *
import numpy as np
class assetAnalyse():#父类
    rtnPerMonth=''#每月收益
    rtnPerYear=''#每年收益
    extraRtnPerMonth=''#每月超额收益
    extraRtnPerYear=''#每年超额收益
    summary=''#汇总
    extraSummary=''#特殊指标汇总
    def __init__(self,assetsPoolType,assetsPool,period,benchmark,factor,groupCount,window):
        if assetsPoolType not in ['dynamic','static']:
            raise Exception('资产池类型输入错误，必须是dynamic或static')
        if assetsPoolType=='dynamic' and not isinstance(assetsPool,str):
            raise Exception('动态资产池必须为字符串')
        if assetsPoolType=='static' and not isinstance(assetsPool,list):
            raise Exception('静态资产池必须为列表')
        self.assetsPoolType=assetsPoolType#资产池类型，dynamic还是static
        self.assetsPool=assetsPool#资产池
        if not isinstance(period,tuple):
            raise Exception('起止日期必须是tuple')
        self.start_date=period[0]#开始日期
        self.end_date=period[1]#结束日期
        self.monthStart=[]#每月第一个交易日
        self.yearStart=[]#每年第一个交易日
        self.tradeDays=[]#测试区间每一个交易日
        self.factor=factor#分析因子,dict类型，{'因子':boolean,True为升序，False为降序}
        self.window=window#因子取值时间
        self.groupCount=groupCount # 分组数
        self.benchmark=benchmark #基准指数
        self.assets=[]#记录当前资产池
        #获取区间内每个周期起点和最后的终点
        self._get_period_start()
        self.portfolio=''#记录每日净值
        self._group_assets=''#记录当前分组
    #因子计算z值，便于因子值相加    
    def _zscore(self,data):
        mean=data.mean()
        std=data.std()
        data=(data-mean)/std
        return data
    #计算测试区间相关日期：每月、每年第一个交易日，所有交易日。
    def _get_period_start(self):
        cal=DataAPI.TradeCalGet(exchangeCD=u"XSHG",beginDate=self.start_date,endDate=self.end_date,isOpen=u"1",field=u"",pandas="1")
        monthStart=[]
        yearStart=[]
        #将距离起点终点最近的交易日作为起终点
        cal['calendarDate1']=pd.to_datetime(cal['calendarDate'])
        cal['from_start']=(cal['calendarDate1']-datetime.strptime(self.start_date,'%Y%m%d'))
        cal['from_end']=(cal['calendarDate1']-datetime.strptime(self.end_date,'%Y%m%d'))
        cal.set_index('calendarDate1',inplace=True)
        start_day=cal['from_start'].idxmin()
        end_day=cal['from_end'].idxmax()
        #区间所有日期
        in_month=list(cal[cal['isMonthEnd'].shift(1)==1]['calendarDate'])
        in_year=list(cal[cal['isYearEnd'].shift(1)==1]['calendarDate'])
        if len(in_month)>0:
            if in_month[0]!=start_day.strftime('%Y-%m-%d'):
                monthStart.append(start_day.strftime('%Y-%m-%d'))
        else:
            monthStart.append(start_day.strftime('%Y-%m-%d'))
        if len(in_year)>0:
            if in_year[0]!=start_day.strftime('%Y-%m-%d'):
                yearStart.append(start_day.strftime('%Y-%m-%d'))
        else:
            yearStart.append(start_day.strftime('%Y-%m-%d'))
        monthStart.extend(in_month)
        yearStart.extend(in_year)
        if len(in_month)>0:
            if in_month[-1]!=end_day.strftime('%Y-%m-%d'):
                monthStart.append(end_day.strftime('%Y-%m-%d'))
        else:
            monthStart.append(end_day.strftime('%Y-%m-%d'))
        if len(in_year)>0:
            if in_year[-1]!=end_day.strftime('%Y-%m-%d'):
                yearStart.append(end_day.strftime('%Y-%m-%d'))
        else:
            yearStart.append(end_day.strftime('%Y-%m-%d'))
        self.monthStart=monthStart
        self.yearStart=yearStart
        self.tradeDays=list(DataAPI.TradeCalGet(exchangeCD=u"XSHG",beginDate=start_day,
            endDate=end_day,isOpen=u"1",field=u"calendarDate",pandas="1")['calendarDate'])
        return
    #计算因子函数,子类实现
    def _get_factor(self,tradeDate,assets):
        return
    #分组函数
    def _group(self,tradeDate):
        if self.assetsPoolType=='dynamic':#动态资产池，获取当前资产池
            assets=list(DataAPI.IdxConsGet(secID=self.assetsPool,intoDate=tradeDate,field=u"consID",pandas="1")['consID'])
        else:
            assets=self.assetsPool
        self._assets=assets
        #计算因子值，子类实现
        score=self._get_factor(tradeDate)
        #每组资产数量
        per_assets=int(len(assets)/self.groupCount)
        group_assets={}
        #按因子值分组
        for i in range(len(score)):
            tmp=int(i/per_assets)+1
            if tmp>self.groupCount:tmp=self.groupCount
            group_assets[score.index[i]]=tmp
        self._group_assets=pd.Series(group_assets)
        return
    #计算当日净值函数，子类实现
    def _calRtn(self,tradeDate,base_month_pos):
        return
    #汇总函数
    def _summary(self):
        rtn=self.portfolio
        #计算每月收益
        _rtnPerMonth=rtn.loc[self.monthStart]
        self.rtnPerMonth=_rtnPerMonth/_rtnPerMonth.shift(1)-1
        #计算每年收益
        self.rtnPerYear=rtn.loc[self.yearStart]
        self.rtnPerYear=self.rtnPerYear/self.rtnPerYear.shift(1)-1
        #计算超额收益
        tmpDf=self.rtnPerMonth.copy()
        for col in tmpDf.columns:
            tmpDf[col]=tmpDf[col]-tmpDf['benchmark']
        self.extraRtnPerMonth=tmpDf
        tmpDf=self.rtnPerYear.copy()
        for col in tmpDf.columns:
            tmpDf[col]=tmpDf[col]-tmpDf['benchmark']
        self.extraRtnPerYear=tmpDf
        #汇总表
        total=pd.DataFrame(columns=rtn.columns)
        total.loc['总收益']=rtn.iloc[-1]-1
        total.loc['超额收益率']=rtn.iloc[-1]-rtn['benchmark'].iloc[-1]
        self.summary=total
        #需要分别计算的指标
        results={}
        for group in range(1,self.groupCount+1):
            result={}
            #计算胜率
            result['月胜率']=self.extraRtnPerMonth[group][self.extraRtnPerMonth[group]>0].count()/float(len(self.extraRtnPerMonth))
            #年化收益率
            last_rtn=rtn[group].iloc[-1]
            result['年化收益率']=pow(last_rtn,250/float(len(rtn)))-1
            #年化波动率
            result['年化波动率']=rtn[group].std()*np.sqrt(len(rtn))
            #sharpe
            result['夏普比率']=(result['年化收益率']-0.03)/result['年化波动率']
            #最大回撤
            result['最大回撤']=max(1-rtn[group]/rtn[group].cummax())
            results[group]=result
        self.extraSummary=pd.DataFrame(results)
    #主函数
    def analyze(self):
        #初始化变量
        #记录每个分组编号
        groups=[]
        for i in range(1,self.groupCount+1):
            groups.append(i)
        groups.append('mean')#记录资产池平均收益
        groups.append('benchmark')#记录基准收益
        #初始化每日净值
        init_value={}
        for i in groups:
            init_value[i]=1
        self.portfolio=pd.DataFrame(init_value,index=[self.yearStart[0]])
        base_date=self.monthStart[0]#计算收益的基准日
        #第一天分组
        #print(self.tradeDays)
        self._group(self.tradeDays[0])
        #循环每个交易日
        for i in range(1,len(self.tradeDays)):
            tradeDate=self.tradeDays[i]
            self._calRtn(tradeDate,base_date)
            if tradeDate in self.monthStart:
                self._group(tradeDate)
                base_date=tradeDate
        self._summary()        
#基金子类
class fundAnalyse(assetAnalyse):
    #计算因子值
    def _get_factor(self,tradeDate):
        f=self.factor.keys()
        if len(f)==1:
            data=DataAPI.FundPerfIndicGet(dataDate=tradeDate,secID=self.assetsPool,window=self.window,
                field=u"secID,"+f[0],pandas="1")
        else:
            data=DataAPI.FundPerfIndicGet(dataDate=tradeDate,secID=self.assetsPool,window=self.window,
                field=u"secID,"+','.join(f),pandas="1")
        #对每个因子值按设定顺序排列。从大到小的以最低值为基准，反之以最大值为基准。
        for fa in f:
            data[fa]=self._zscore(data[fa])
            if self.factor[fa]==False:#从大到小排列
                data[fa]=data[fa]-data[fa].min()
            else:
                data[fa]=data[fa].max()-data[fa]
        data['score']=data.sum(axis=1)
        data.set_index('secID',inplace=True)
        data.sort_values('score',ascending=False,inplace=True)
        #print(data)
        return data['score']
    #计算收益
    def _calRtn(self,tradeDate,base_date):
        portfolio=self.portfolio
        #获取本月起点净值
        base=portfolio.loc[base_date]#series
        #获取资产池各资产相对周期起点的涨幅
        data=DataAPI.FundNavGet(secID=self._assets,beginDate=base_date,endDate=tradeDate,
            field=u"secID,endDate,ADJUST_NAV",pandas="1")
        data=data.groupby(['secID','endDate']).mean().unstack().T
        grow=data.iloc[-1]/data.iloc[0]
        mean=grow.mean()#资产平均涨幅
        #计算每组平均涨幅
        group_grow=pd.DataFrame(index=self._group_assets.index)
        group_grow['nav']=grow
        group_grow['group']=self._group_assets
        #print tradeDate
        #print base_date
        group_grow=group_grow.groupby('group').mean()
        #获取基准指数涨幅
        benchmark_nav=DataAPI.MktIdxdGet(indexID=self.benchmark,beginDate=base_date,endDate=tradeDate,
            field=u"closeIndex",pandas="1")
        benchmark_grow=benchmark_nav.iloc[-1]/benchmark_nav.iloc[0]
        #拼接所有涨幅
        group_grow.loc['mean']=mean
        group_grow.loc['benchmark']=benchmark_grow.values[0]
        #计算当前净值
        curr_nav=base.multiply(group_grow['nav'])
        self.portfolio.loc[tradeDate]=curr_nav
        #print self._portfolio
        return
```

# 以下用一个简单的双因子做测试。对部分场内ETF进行分析。

```python
assetPool=['510050.XSHG','510300.XSHG','510500.XSHG', '159915.XSHE','159905.XSHE','159928.XSHE','512010.XSHG','159940.XSHE']
#因子选取年化收益和TM模型的选股能力。
factor={}
factor['annTotReturn']=False
factor['selectionTM']=False
result={}
extraResult={}
#分别对1月、半年、1年、3年周期的表现进行测试
for window in ['1','3','5','7']:
    aa=fundAnalyse('static',assetPool,('20190101','20210801'),'000300.ZICN',factor,4,window)
    aa.analyze()
    result[window]=aa.summary
    extraResult[window]=aa.extraSummary
```

# 分析结果。可以看出以半年窗口的因子收益最高，且每组收益单调递减。
