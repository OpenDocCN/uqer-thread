# 巨潮指数分析

巨潮系列风格指数提供了大中小盘、价值、成长几种风格的指数，可以作为观察市场风向的基准。本文对巨潮系列指数进行分析和对比。

```python
from lib.functions import *
import matplotlib.pyplot as plt
```

# 首先定义一个获得指定时间段收益并绘制曲线的函数。
参数index:要分析的指数代码ticker
period:列表，第一个元素是起始时间，第二个是结束时间。
plot:是否绘制曲线

```python
# 指定日期段收益分析
def ret_by_period(index,period,plot=False):
    #获取收盘点位
    price=DataAPI.MktIdxdGet(ticker=index,
                         beginDate=period[0],endDate=period[1],field=u"ticker,tradeDate,closeIndex",pandas="1")
    #将交易日期转为日期格式
    price['tradeDate']=pd.to_datetime(price['tradeDate'],format='%Y-%m-%d')
    #转为代码-日期表格
    price=price.groupby(['ticker','tradeDate']).mean().unstack().T
    #计算每日相对起始日收益
    price=price/price.iloc[0]-1
    price=price.loc['closeIndex']
    #绘制曲线
    if plot:
        plt.figure(figsize=[15,10],dpi=200)
        for c in price.columns:
            plt.plot(price[c],label=c)
        plt.legend()
    return price.iloc[-1]
```
