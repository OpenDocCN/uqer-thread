```python
###g1maxr涨幅最大,g2liuchang走势流畅
####1.初始设置：导入模块、设置日期、股票池等
import pandas as pd
import datetime as dt
universe=set_universe('A')
xingu=DataAPI.EquGet(secID=universe,ticker=u"",listStatusCD=u"",field=['secID','secShortName','listDate'],pandas="1")
xingu=xingu.ix[xingu.listDate>'2016-11-23',:]
for stk in xingu.secID:
    if stk in universe:
        universe.remove(stk)
# universe=set_universe('HS300')#先以沪深300为股票池

# 抓取最低价的日期区间
end='20170228'
begin='20170113'
list=[]
lowestlist=pd.DataFrame()
```

```python
####2.在设定时间段内，获取股票池所有股票的日行情信息,存在price里  
# for stock in universe:
#     price=DataAPI.MktEqudAdjGet(secID=stock,beginDate=begin,endDate=end,field=['ticker','secShortName','tradeDate','lowestPrice','closePrice'],pandas='1')
#     # price.index=price['ticker']
#     # price.drop('ticker',axis=1,inplace=True)

# ## 储存最低价及其最低价出现的交易日期,存在lowestlist
#     if len(lowestlist)==0:
#         lowestlist=price[price.lowestPrice==price.lowestPrice.min()]
#     else:
#         lowestlist=lowestlist.append(price\
#                                      [price.lowestPrice==price.lowestPrice.min()],ignore_index=True)
# lowestlist.drop_duplicates(subset='ticker',keep='last',inplace=True)

####优化加速版
price=DataAPI.MktEqudAdjGet(secID=universe,beginDate=begin,\
                            endDate=end,field=['ticker','secShortName','tradeDate',\
                                               'lowestPrice','closePrice'],pandas='1')
price.sort(columns=['ticker','lowestPrice'],ascending=[True,True],inplace=True)
lowestlist=price.drop_duplicates(['ticker'],keep='first')
lowestlist.index=range(0,len(lowestlist))
```

```python
####3.借鉴最大回撤指标，计算上述股票在给定区间的最大回撤，并存入infor。
infor=lowestlist.ix[:,0:4]
infor['drawback']=0
today=dt.datetime.today().strftime('%Y%m%d')##先转为Timestamp，再转为yyyymmdd字符型日期
for i in infor.index:
    # date=pd.to_datetime(lowestlist.ix[i,'tradeDate']).strftime('%Y%m%d')
    # data=DataAPI.MktEqudAdjGet(ticker=lowestlist.ix[i,'ticker'],beginDate=date,endDate=today,\
    #                            pandas='1',field=['ticker','secShortName','tradeDate','closePrice'])
    data=DataAPI.MktEqudAdjGet(ticker=lowestlist.ix[i,'ticker'],\
                               beginDate=pd.to_datetime(lowestlist.ix[i,'tradeDate'])\
                               .strftime('%Y%m%d'),endDate=end,pandas='1',\
                               field=['ticker','secShortName','tradeDate','highestPrice',\
                                      'lowestPrice','closePrice'])
    data['high']=0
    data['huiche']=0

    for j in range(0,len(data.tradeDate),1):
        # lowdate=lowestlist.ix[j,'tradeDate']
        data.ix[j,'high']=data.ix[data.tradeDate<data.ix[j,'tradeDate'],'highestPrice'].max()
        high=data.ix[data.tradeDate<data.ix[j,'tradeDate'],'highestPrice'].max()
        # print (high)
        # print (data.ix[i,3])
        data.ix[j,'huiche']=1-(data.ix[j,'lowestPrice']/high)
    infor.ix[i,'drawback']=data.ix[:,'huiche'].max()
        
```

```python
####4.处理区间最大涨幅
infor['return']=0
infor['maxre']=0
for i in infor.index:
    data2=DataAPI.MktEqudAdjGet(ticker=lowestlist.ix[i,'ticker'],\
                               beginDate=pd.to_datetime(lowestlist.ix[i,'tradeDate'])\
                                .strftime('%Y%m%d'),endDate=end,\
                               pandas='1',field=['ticker','tradeDate','closePrice'])
    a=len(data2)-1
    maxp=data2.closePrice.max()
    minp=infor.ix[i,'lowestPrice']
    maxre=(maxp/minp)-1
    re=(data2.ix[a,'closePrice']/minp)-1
    infor.ix[i,'maxre']=maxre
    infor.ix[i,'return']=re
```

```python
####5.后续处理
infor['r/d']=infor.maxre/infor.drawback
infor['re/d']=infor['return']/infor.drawback

result=infor.ix[infor.lowestPrice!=0,:]
# result1=result.ix[result.drawback<=0.1,:]
# result2=result1.ix[result1.maxre<=1,:]
result2=result.ix[result.drawback>0,:]
# result4=result3.sort(columns=['drawback','maxre','return'],axis=0,ascending=[True,False,False])
g2liuchang=result2.sort(columns=['re/d'],axis=0,ascending=[False])
# g2liuchang=result2.sort(columns=['drawback','return','maxre'],axis=0,ascending=[True,False,False])
# g2liuchang=g2liuchang.ix[g2liuchang.tradeDate<='2016-12-15',:]
g1max=result2.sort(columns=['return'],axis=0,ascending=[False])
# result5=result4.ix[result4.tradeDate<='2017-01-31',:]
# result5
```
