# OU过程（均值回归过程）

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
from numpy import random
r0=0.02
time=100
delta=1.0/252
sigma=0.03

ou_a=20 #rate of mean reversion

ou_mu=0.02 #long run average

end=90 #time T

end=end+1

i=500 #模拟次数

barrier=0.01
#print(random.normal(loc=0,scale=sqrt(1.0/252)*sigma ,size=10))
def brownian_motion_logretunrs(dt,sig):
    sqrt_delta_sigma=math.sqrt(dt)*sig
    return random.normal(loc=0,scale=sqrt_delta_sigma,size=time)

 
def ornstein_uhlenbeck(t,a,mu,dt):
    paths=[r0]
    brownian_motion_returns=brownian_motion_logretunrs(delta,sigma)
    for i in range(1,t):
        drift=a*(mu-paths[i-1])*dt
        randomness=brownian_motion_returns[i-1]
        paths.append(paths[i-1]+drift+randomness)
    return paths


paths=[]
for i in range(i):
    level=ornstein_uhlenbeck(time,ou_a,ou_mu,delta)
    paths.append(level)
    
paths=np.asarray(paths)
paths=paths.T
new_paths=np.delete(paths,np.s_[end:],0)

print(np.shape(new_paths))
 
plt.plot(new_paths,lw=0.5)
plt.axhline(y=barrier,lw=0.7,color='b')
plt.ylabel('level')
plt.show()
 
```
