```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
start= '2017-01-01'
end = '2017-12-31'
global TradeCal #将交易所日历表设置为全局变量
TradeCal = DataAPI.TradeCalGet(exchangeCD=u"XSHG",beginDate=start,endDate=end,field=['calendarDate','isOpen'],pandas="1")
TradeCal = TradeCal[TradeCal.isOpen==1][['calendarDate']] #排除非交易日,只取日历列表

####################################################################
def tradeDays(date0,date1): #返回两个日期间的交易天数，排除了休市日
    global TradeCal #交易所日历
    if date0>date1:
        minDate = date1
        maxDate = date0
    else:
        minDate = date0
        maxDate = date1
    days = TradeCal[(TradeCal.calendarDate>=minDate) & (TradeCal.calendarDate<=maxDate)].count()
    return days
#end def
```
