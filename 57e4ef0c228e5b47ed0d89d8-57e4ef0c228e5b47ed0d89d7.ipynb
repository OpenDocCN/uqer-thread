短线交易
5日均线， 10日均线突破法
策略逻辑：
1.5日均线向上突破10日均线，若当前有空仓，平掉，若当前多仓为0,买入1手
2.5日均线向下突破10日均线，若当前空仓为0，卖出1手，若当前有多仓，平掉

```python
import talib
import pandas as pd
import numpy as np

import quartz_futures as qf
from quartz_futures.api import *

universe = ['IH1609']

start = "2016-03-01"                 # 回测开始时间
end   = "2016-09-15"                 # 回测结束时间
capital_base = 1000000               # 初始可用资金
refresh_rate = 1                     # 算法调用周期
freq = 'd'
commission = 0.3/10000
margin = 0.3
multiplier = 300


def initialize(futures_account):
    pass

def handle_data(futures_account):
    symbol = 'IH1609'
    print futures_account.current_date, futures_account.current_time, symbol
    print futures_account.position
    print futures_account.position.keys()
    data1 = get_symbol_history(symbol=symbol, field='closePrice', time_range=5)
    data2 = get_symbol_history(symbol=symbol, field='closePrice', time_range=10)
    closeprice_5 = np.array(data1[symbol]['closePrice'], dtype='float')
    closeprice_10 = np.array(data2[symbol]['closePrice'], dtype='float')
    MA5 = talib.MA(closeprice_5, 5)[-1]
    MA10 = talib.MA(closeprice_10, 10)[-1]
    print '\n5日均线：', MA5, '10日均线', MA10, '昨日收盘', closeprice_5[-1]
    if MA5 > MA10:
        long_position = futures_account.position.get(symbol, dict()).get('long_position', 0)
        short_position = futures_account.position.get(symbol, dict()).get('short_position', 0)
        if long_position == 0 :
            print futures_account.current_date, futures_account.current_time, '买入开仓'
            order(symbol, 1, 'open')
        if short_position > 0:
            print futures_account.current_date, futures_account.current_time, '买入平仓'
            order(symbol, 1, 'close')
    if MA5 <= MA10:
        long_position = futures_account.position.get(symbol, dict()).get('long_position', 0)
        short_position = futures_account.position.get(symbol, dict()).get('short_position', 0)
        if long_position > 0:
            print futures_account.current_date, futures_account.current_time,'卖出平仓'
            order(symbol, -1, 'close')
        if short_position == 0:
            print futures_account.current_date, futures_account.current_time, '卖出开仓'
            order(symbol, -1, 'open')
bt, perf = qf.backtest.backtest(universe=universe, start=start, end=end,
                                initialize=initialize, handle_data=handle_data,
                                capital_base=capital_base, refresh_rate=refresh_rate,
                                freq=freq)
bt
```
