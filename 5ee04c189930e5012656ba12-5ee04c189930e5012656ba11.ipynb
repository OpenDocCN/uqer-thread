```python
import numpy as np
import pandas as pd
import scipy.stats as st
```

```python
benchmark = 'HS300'
date='20171229'
universe =set_universe(benchmark,date)
factor = 1/DataAPI.MktStockFactorsOneDayGet(tradeDate=date,secID=universe,\
                                          field=u"secID,PE",pandas="1").set_index('secID').dropna()
factor.columns=['EP']
```

```python
#前50等分
signal_1 = factor.sort_values(by='EP',ascending = False).head(50)
signal_1['weight']=1/len(signal_1)
signal_1
```

```python
#符合条件的，按照EP权重
signal_2 = factor[factor['EP']>0]
signal_2['weight']=signal_2['EP']/signal_2['EP'].sum()
signal_2
```

```python
#大于一定分位数的，等分
quantile_value = factor['EP'].quantile(0.9)
signal_3 = factor[factor['EP']>quantile_value]
signal_3['weight']=1/len(signal_3)
signal_3
```

```python
month_trade_list=DataAPI.TradeCalGet(exchangeCD=u"XSHG",\
                                       endDate=date,isOpen=u"1",field=u"calendarDate,isMonthEnd",pandas="1")
month_trade_list=month_trade_list[month_trade_list['isMonthEnd']==1]
last_month_date =month_trade_list.iloc[-1,0]
last_month_date
```

```python
benchmark_weight = DataAPI.IdxCloseWeightGet(secID=u"",ticker=u"000300",beginDate=last_month_date\
                                             ,endDate=last_month_date,field=u"consID,consShortName,weight",pandas="1")
benchmark_weight.head()
```

```python
industry = DataAPI.EquIndustryGet(industryVersionCD=u"010303",\
                                  intoDate=last_month_date,field=u"secID,secShortName,industryName1",pandas="1")
industry.head()
```

```python
benchmark_weight = pd.merge(benchmark_weight,industry,left_on ='consID',right_on ='secID')
benchmark_weight.head()
```

```python
benchmark_industry_weight = benchmark_weight.groupby('industryName1').sum()
benchmark_industry_weight = benchmark_industry_weight/benchmark_industry_weight.sum()
benchmark_industry_weight.columns = ['industry_weight']
benchmark_industry_weight
```
