## 2. welcome the cache function

```python
def wrapper_func_cache(func):
    @wraps(func)
    def wrapped(*args, **kwargs):
        if not hasattr(func, '__cache_result'):
            func.__cache_result = {}
        
        hash_key = [str(i) for i in args] + [str(j) for j in kwargs.iteritems()]
        cache_key = hash(tuple(hash_key))
        
        if cache_key not in func.__cache_result:
            func.__cache_result[cache_key] = func(*args, **kwargs)
            
        return func.__cache_result[cache_key]
    return wrapped
```

## 3. test everything you write

```python
@wrapper_func_cache
def MktEquFlowOrderGetCache(secID, beginDate, endDate):
    """MktEquFlowOrderGet cache version.
    """
    return DataAPI.MktEquFlowOrderGet(secID=secID, beginDate=beginDate, endDate=endDate)

DataAPI.MktEquFlowOrderGetCache = MktEquFlowOrderGetCache
```

### 4.1 策略逻辑：[『小试牛刀』个股资金流](https://uqer.datayes.com/community/share/5696099e228e5b18dfba2c8b)

```python

start = '2015-11-01'                       # 回测起始时间
end = '2016-01-01'                         # 回测结束时间
benchmark = 'HS300'                        # 策略参考标准
universe = set_universe("A")               # 证券池，支持股票和基金
capital_base = 100000                      # 起始资金
freq = 'd'                                 # 策略类型，'d'表示日间策略使用日线回测，'m'表示日内策略使用分钟线回测
slippage = Slippage(value=0.0)
refresh_rate = 4                           # 调仓频率，表示执行handle_data的时间间隔，若freq = 'd'时间间隔的单位为交易日，若freq = 'm'时间间隔为分钟


def my_filter(tradeDate, preTradeDate):
    data = DataAPI.MktEquFlowOrderGet(secID=set_universe("A"), beginDate=preTradeDate, endDate=preTradeDate)
    data = data[(data.netInflowS > 0) & (data.netInflowM >= data.netInflowS) & 
                (data.netInflowL >= data.netInflowM) & (data.netInflowXl >= data.netInflowL)]
    return data

def my_filter_cache(tradeDate, preTradeDate):
    data = DataAPI.MktEquFlowOrderGetCache(secID=set_universe("A"), beginDate=preTradeDate, endDate=preTradeDate)
    data = data[(data.netInflowS > 0) & (data.netInflowM >= data.netInflowS) & 
                (data.netInflowL >= data.netInflowM) & (data.netInflowXl >= data.netInflowL)]
    return data

def initialize(account):                   # 初始化虚拟账户状态
    pass

def handle_data(account):                  # 每个交易日的买入卖出指令
    tradeDate = account.current_date.strftime('%Y-%m-%d')
    preTradeDate = lib.const_pre_trade_day(tradeDate)
    
    data = my_filter(tradeDate, preTradeDate)
    
    ### position manage
    def clear_position():
        for secID in account.valid_secpos:
            order_to(secID, 0)
            
    def open_position():
        for secID in data.secID:
            if account.referencePrice[secID] > 0:
                volume = int(account.referencePortfolioValue/len(data)/account.referencePrice[secID]/100) * 100
                order(secID, volume)

    clear_position()
    open_position()

    ### debug info
    def info():
        print '-' * 80
        print 'tradeDate: {}, preTradeDate: {}'.format(tradeDate, preTradeDate)
    # info()
    
    return

def handle_data_cache(account):                  # 每个交易日的买入卖出指令
    tradeDate = account.current_date.strftime('%Y-%m-%d')
    preTradeDate = lib.const_pre_trade_day(tradeDate)
    
    data = my_filter_cache(tradeDate, preTradeDate)
    
    ### position manage
    def clear_position():
        for secID in account.valid_secpos:
            order_to(secID, 0)
            
    def open_position():
        for secID in data.secID:
            if account.referencePrice[secID] > 0:
                volume = int(account.referencePortfolioValue/len(data)/account.referencePrice[secID]/100) * 100
                order(secID, volume)

    clear_position()
    open_position()

    ### debug info
    def info():
        print '-' * 80
        print 'tradeDate: {}, preTradeDate: {}'.format(tradeDate, preTradeDate)
    # info()
    
    return
```

`测试版本1`

- 掉用老版 DataAPI.MktEquFlowOrderGet 获取个股资金流数据
	+ 理想情况：回测速度慢

```python
tstart = time.time()

# 把回测参数封装到 SimulationParameters 中，供 quick_backtest 使用
sim_params = quartz.SimulationParameters(start, end, benchmark, universe, capital_base)

# 获取回测行情数据
idxmap, data = quartz.get_daily_data(sim_params)

# 运行结果
results = {}

# 把回测逻辑封装到 TradingStrategy 中，供 quick_backtest 使用
strategy = quartz.TradingStrategy(initialize, handle_data)

# 回测部分
bt, acct = quartz.quick_backtest(sim_params, strategy, idxmap, data, refresh_rate = refresh_rate)

# 对于回测的结果，可以通过 perf_parse 函数计算风险指标
perf = quartz.perf_parse(bt, acct)

# 保存运行结果
results['returns'] = perf['returns']

# 转换为 DataFrame 并画图
import pandas
results = pandas.DataFrame(results)
results.plot(figsize=(20, 6)) 

tend = time.time()

print 'consume {} seconds'.format(tend - tstart)
```

`相关 lib 代码`

```python
############################################
### constants
############################################

@wrapper_func_cache
def const_trade_calendar():
    data = DataAPI.TradeCalGet(exchangeCD=u"XSHG,XSHE", beginDate=u"20050101", endDate=u"", field=['calendarDate', 'isOpen'])
    data = data[data.isOpen == 1]

    return list(data.calendarDate)


@wrapper_func_cache
def const_pre_trade_day(day):
    trade_days = const_trade_calendar() 
    index = trade_days.index(day)
    
    return trade_days[index - 1]
```
