# **概览**
***
优化是一个非常麻烦的事情，我们常常被坑于此，好在python里scipy中的optimize子包中提供了常用的最优化算法函数实现，可以帮助我们处理很多优化的问题。

我们从最基本的`单变量非线性最优化问题`开始，对比一下各个算法的优劣，下面是optimize所提供的处理单变量非线性最优化问题的常用函数：

- fmin  --  简单Nelder-Mead算法
- fmin_powell --  改进型Powell法
- fmin_bfgs  --  拟Newton法
- fmin_cg -- 非线性共轭梯度法
- fmin_ncg  --  线性搜索Newton共轭梯度法

# **对比**
***
- 目标函数
	
    我们不妨取一个简单的二次函数$x^2-6x+6$作为对比的对象,它的最小值在$x=3$处取得，最小值为-3，下面先定义这个目标函数

```python
import numpy as np
import pandas as pd
import scipy.optimize as opt
import matplotlib
import matplotlib.pyplot as plt
```

```python
#目标函数
def objectfun(x):
    return x**2-6*x+6
```

- **fmin -- 简单Nelder-Mead算法：**

```python
x0 = 0  #设置一个初值
opt.fmin(objectfun, x0)
```

这个结果还是非常准确的，我们不妨设置一系列初值来看看对结果有没有影响：

```python
result1 = pd.DataFrame()
x0 = np.arange(30)*0.2
result1['初值'] = x0
result1['最小值点'] = [opt.fmin(objectfun, x0[i])[0] for i in range(len(x0))]
```

```python
result1.T
```

```python
(result1['最小值点']-3).plot(figsize=[12,6])
```

可以看到，随着初值点选取得不同，最小值点的值在3附近波动了，这是为什么呢？可能是二进制编码的问题，也可能是算法本身的问题。

这个问题还比较简单，复杂些的多元情况下，不知道会不会造成很大的影响~

- **fmin_powell -- 改进型Powell法**
	
    同样的思路把其它的结果算出来对比一下：

```python
result2 = pd.DataFrame()
x0 = np.arange(30)*0.2
result2['初值'] = x0
result2['最小值点'] = [opt.fmin_powell(objectfun, x0[i]) for i in range(len(x0))]
```

```python
result2.T
```

- **fmin_bfgs  --  拟Newton法**

```python
result3 = pd.DataFrame()
x0 = np.arange(30)*0.2
result3['初值'] = x0
result3['最小值点'] = [opt.fmin_bfgs(objectfun, x0[i])[0] for i in range(len(x0))]
```

```python
result3.T
```

- **fmin_cg -- 非线性共轭梯度法**

```python
result4 = pd.DataFrame()
x0 = np.arange(30)*0.2
result4['初值'] = x0
result4['最小值点'] = [opt.fmin_cg(objectfun, x0[i])[0] for i in range(len(x0))]
```

```python
result4.T
```

- **fmin_ncg  --  线性搜索Newton共轭梯度法**

```python
#梯度向量
def Gradient(x):
    return 2*x-6
```

```python
result5 = pd.DataFrame()
x0 = np.arange(30)*0.2
result5['初值'] = x0
result5['最小值点'] = [opt.fmin_ncg(objectfun, x0[i],fprime=Gradient)[0] for i in range(len(x0))]
```

```python
result5.T
```

# **总结**
***

```python
result = pd.DataFrame()
result['简单Nelder-Mead算法'] = result1['最小值点']
result['改进型Powell法'] = result2['最小值点']
result['拟Newton法'] = result3['最小值点']
result['非线性共轭梯度法'] = result4['最小值点']
result['线性搜索Newton共轭梯度法'] = result5['最小值点']
result
```

事实告诉我们，在这个问题上，拟Newton法，非线性共轭梯度法，线性搜索Newton共轭梯度法的结果比较好，也比较稳定。梯度算法是个好东西。更多优化问题和方法参见大神薛昆的帖子[量化分析师的Python日记【第4天：一大波金融Library来袭之scipy篇】](https://uqer.datayes.com/community/share/54d83bb3f9f06c276f651a6e)
