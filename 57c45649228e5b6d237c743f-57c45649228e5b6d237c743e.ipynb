-------
## 因子的预测能力
> 直接计算Rank-IC（频率为周），并且求其20周的移动均值。如果Rank-IC均线（红色曲线）分布在X轴（均值为0）下方或者上方，证明该因子有较稳定的预测性。反之则稳定性不强。

-------
## 技术实现
> 从速度上考虑，对原来的代码进行了稍微的调整，即每次调用 DataAPI.MktStockFactorsOneDayGet 函数时直接一次取出所有的因子数据，然后再对每个因子计算 Rank-IC 值，这样避免了多次调用 DataAPI 函数的时间开销。


```python
import scipy.stats as st
import pandas as pd

DataAPI.settings.cache_enabled = True 

def get_weekend_trading_day_list(begin_day="20140101", end_day="20160801"):
    """获取每周的最后一个交易日"""
    cal_dates = DataAPI.TradeCalGet(exchangeCD=u"XSHG", beginDate=begin_day, endDate=end_day, field="calendarDate,isWeekEnd")
    trading_days = cal_dates[cal_dates['isWeekEnd']==1]["calendarDate"].tolist()
    trading_days = [day[0:4] + day[5:7] + day[8:] for day in trading_days]  # 更改日期的格式，将'2014-01-03'转化为'20140103'
    return trading_days

def get_index_cons(day="20140110", ticker="000300"):
    """获取当日指数的成份股，结果用字符串拼接起来方便后续的传参"""
    cons_id_df = DataAPI.IdxConsGet(secID=u"",ticker=ticker,intoDate=day,isNew=u"",field=u"consID",pandas="1")
    cons_id_ls = cons_id_df["consID"].tolist()
    return cons_id_ls

def plot_result(result_df, factor_name, ticker):
    """可视化最后的结果"""
    import matplotlib
    import matplotlib.pyplot as plt
    matplotlib.style.use('ggplot')
    result_df["date"] = result_df["date"].apply(str)
    result_df.index = pd.to_datetime(result_df["date"])
    result_df.plot(colors=['royalblue', 'crimson']).set_title(factor_name+" in "+ticker)
    plt.axhline(0, color='Gray')

def get_all_factors():
    df = DataAPI.MktStockFactorsOneDayGet(tradeDate='20161020', secID='600000.XSHG')
    cols = sorted(df.columns.tolist())
    for col in ['secID', 'ticker', 'tradeDate']:
        cols.remove(col)
    return cols

def make_factors_rankic(factor_names, index="000300", begin_day="20140101", end_day = "20160801"):
    trading_days = get_weekend_trading_day_list(begin_day, end_day)
    if factor_names is None:
        factors = get_all_factors()
    else:
        assert(factor_names.strip() != '')
        factors = factor_names.split(',')
    #print('factors: %d' % len(factors))
    ic_ls = [] #保存计算结果，最终转化为DataFrame格式
    for i in xrange(len(trading_days)-1):
        curr_day = trading_days[i]
        next_day = trading_days[i+1]
        sons_id_str = get_index_cons(curr_day, index)
        factors_df = DataAPI.MktStockFactorsOneDayGet(tradeDate=curr_day, secID=sons_id_str) #获取每周最后一个交易日的因子值
        weekly_return = DataAPI.MktEquwAdjGet(secID=sons_id_str, beginDate=next_day, endDate=next_day, field=u"secID,return") # 获取相应股票未来一周的收益
        result = factors_df.merge(weekly_return, on='secID', how="inner")
        mask = ~(abs(result["return"]) < 1e-4)
        result = result.loc[mask]
        factors_ic = {'date' : curr_day}
        for factor_name in factors:
            ic, p_value = st.pearsonr(result[factor_name].rank(),result["return"].rank())   #计算Rank-IC
            factors_ic[factor_name] = ic
        ic_ls.append(factors_ic)
    df = pd.DataFrame(ic_ls)
    return df
```

## 来做个简单的测试

## 结果的直观显示
> 可以看到结果中包含了很多无效的值，另外虽然我们已经有了所有因子的 Rank-IC 值，但是还需要进一步进行处理来方便对比观测。

### 转换成百分数
> 我们知道如果 Rank-IC 值长期保持在零值上方或者下方，表示该因子具有更强的预测能力。那么我们可以统计该因子的 Rank-IC 值大于零的次数以及它所占的百分比，偏离50%越远的百分比就可以表示更强的预测能力。



```python
def rankic_to_percent(factors_df):
    df = factors_df[factors_df.columns[~pd.isnull(factors_df).all(axis=0)]] # 首先过滤掉一些因子值全为NaN的列
    lt_cnt = df[df.lt(0, axis=0)].count(axis=0) # 按列统计小于零的次数
    gt_cnt = df[df.gt(0, axis=0)].count(axis=0) # 按列统计大于零的次数
    eq_cnt = df[df.eq(0, axis=0)].count(axis=0) # 按列统计等于零的次数
    all_cnt = df.count(axis=0) # 按列统计总个数
    z = pd.concat([gt_cnt, eq_cnt, lt_cnt, all_cnt], axis=1) # 合并
    z.columns = ['gt', 'eq', 'lt', 'all'] # 重新设置列名
    z['percent'] = z['gt'] / z['all'] # 建立一个百分比列
    z.loc['date'] = None # 去掉最后一行
    z = z[z['all'] > 20] # 过滤掉总次数小于等于20的因子
    z.dropna(inplace=True) # 
    z = z.sort('percent') # 按百分比列进行排序
    return z

factors_df = make_factors_rankic(None, index='000300', begin_day='20130101', end_day='20160829')
z = rankic_to_percent(factors_df)
print(z.to_html())
```

## 总结
> 可以看到 Aroon, AroonUp, DIFF, DEA, PSY 因子的 Rank-IC 值最小，ILLIQUIDITY, DASREV, DAREV, AroonDown 因子的 Rank-IC 值最大，它们都具有比较强的因子预测能力。

## 进一步思考
> 由于使用不同的参数会得到不同的计算结果，反映在排序上可能有些因子会忽大忽小，因此考虑因子预测能力的稳健性，我们还需要使用不同的参数进行计算对比，只有能够保持足够稳健的因子才可能真正具有比较强的因子预测能力。

## 使用图示
> 当然最直观的还是图示，我们也可以和 [国海证券《因子预测能力初探》uqer实现（一）](https://uqer.datayes.com/community/share/57a3ea90228e5b9b8fa86b0f) 中一样，通过绘图的方式进行观察。


```python
def draw_factors_rankic(factors_df, window=20, min_periods=10):
    import matplotlib
    import matplotlib.pyplot as plt    
    #df = factors_df.dropna(axis=1)
    df = factors_df
    df['date'] = df['date'].apply(str)
    df.index = pd.to_datetime(df['date']).dt.date
    factors = df.columns.tolist()
    factors.remove('date')
    for idx, factor_name in enumerate(factors):
        factor = df[factor_name]
        fmean = pd.rolling_mean(factor, window=window, min_periods=min_periods)
        result = pd.concat([factor, fmean], axis=1)
        result.columns = [factor_name, 'factor_mean']
        result.plot(colors=['royalblue', 'crimson'])
        plt.axhline(0, color='Gray')
        #if idx == 20: #
        #    break
            
df = factors_df[factors_df.columns[~pd.isnull(factors_df).all(axis=0)]] # 首先过滤掉一些因子值全为NaN的列
all_cnt = df.count(axis=0) # 按列统计总个数
cols = df.columns[all_cnt > 100] # 找出有效值个数大于100的列
x = df[cols] # 过滤掉缺失值太多的列
#print(all_cnt.shape, df.shape, cols.shape, x.shape)
draw_factors_rankic(x) # 显示
```
