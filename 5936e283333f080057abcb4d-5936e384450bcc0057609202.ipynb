# 大市值突破买入

### 小市值股票通常具有较高的收益率，大小盘轮动的策略验证过小盘股整体的涨幅强于大盘股。主要是因为小盘股市值小，资金量不需要太大就能影响股价的变动，这也意味着高收益对应高风险。为了追求策略的稳健性，本策略选择大盘股作为投资标的，尽量控制策略的回撤。具体思路从以下三个步骤介绍：
----
> 1.选股。参考Fama-French三因子模型和论坛多因子模型的研究成果，最终确定使用市值、市盈率、营业收入增长率三个因子筛选股票入选备选股票池。定义一个买入函数对备选股票池进行操作，买入函数主要作用是筛选股价突破150天周期最高点的股票，并根据市盈率和权益回报率综合评价筛选出排序前三的股票。
> 2.择时。按日进行调仓，股票进入股票池后次日执行买入操作。另外加入止盈止损的卖出机制，在个股涨幅大于15%或者跌幅大于7%时执行卖出操作。
> 3.仓位管理。为了分散风险，避免单只股票持有太多，将账户资金等权分为5份购买股票池股票，单支股票所占仓位约为20%。

结论与思考
代码在andsenhan的基础上进行参数微调和修改，修正后的策略将回撤控制到11.2%，从15年开始回测到今天，按该策略的选股和择时机制，完美规避熊市造成的损失。
另外，该策略也存在可提升空间，账户最多持有5支股票，如果在满仓情况下有新的股票入选股票池并不会执行买入操作，而新入选股票池的表现一般都比较好，所以如果增加一个持有股票的淘汰机制可能会增加更多的盈利能力。
