## 那什么是类和对象呢

类是对象的类型，不同于一般的数据类型，类具有行为。类类型可以用来描述一个事物，所以说类可以称为具有相同属性和行为的集合，它们的内部规律和性质相似。类在建立的时候是抽象的，在使用时刻具体化。面向对象就是通过建立类来管理对象。一般类分为新式类和旧式类，举个例子。

```python
##旧式类
class data():
    print "i am a data."
##新式类
class data1(object):
    print " i am another data."
    
a = data()
a1 = data1()
```

创建完了类之后，我们就要具体化实例。通过具体化之后，我们可以通过方法来访问它的一些属性。方法就是与实例绑定的函数，和普通函数不同，方法可以直接访问实例的数据；通过在实例上调用方法，我们就直接操作了对象内部的数据，但无需知道方法内部的实现细节。和静态语言不同，Python允许对实例变量绑定任何数据。但是类的属性分为私有和公有，一般来说，单下划线开头的属性和方法表示protected类型，即只能允许其本身与子类进行访问双下划线开头的属性和方法表示私有类型的变量，只能是允许这个类本身调用进行访问，子类也不可以。这类属性在运行时属性名会加上单下划线和类名。

```python
##实例化
class student():
    def __init__(self, name, id):
        self.name = name
        self.id = id
 
    def getStudent(self):
        print "Name : ", self.name,  "Id: ", self.id
stu1 = student("Jack", 1001)
stu1.getStudent()
```

```python
##私有属性
class Data():
    def __init__(self,data):
        self.__data = data
s2 = [1,2,3]
d2 = Data(s2)
##d2.__data ##因为是私有，所以会报错
print d2._Data__data  ## 但可以通过object._className__attrName 访问私有属性或定义方法
```

面向对象的编程带来的主要好处之一是代码的重用，实现这种重用的方法之一是通过继承机制。继承完全可以理解成类之间的类型和子类型关系。

在Python中继承中的一些特点：
    1.在继承中基类的构造（__init__()方法）不会被自动调用，它需要在其派生类的构造中亲自专门调用。
    2.在调用基类的方法时，需要加上基类的类名前缀，且需要带上self参数变量。区别于在类中调用普通函数时并不需要带上self参数
    3.Python总是首先查找对应类型的方法，如果它不能在派生类中找到对应的方法，它才开始到基类中逐个查找。
如果在继承元组中列了一个以上的类，那么它就被称作"多重继承" 。

```python
class Person(object):
    sex = 'male'
    def __init__(self, s1, s2):
        self.IDcard = s1
        self.name = s2
        print ('__init__(self) of Person')

    def hello(self, friend):
        print 'hello, ', friend, '!'

class Student(Person):
    def __init__(self, num):
        self.number = num

    def fun(self):
        print self.IDcard, self.name, self.number, Student.sex

class Teacher(Person):
    def __init__(self, t, s1, s2):
        self.title = t
        super(Teacher, self).__init__(s1, s2)
        #Person.__init__(self, s1, s2)
    def fun(self):
        print self.IDcard, self.name, self.title, Teacher.sex


stud1 = Student('2012009')
stud1.IDcard = '00000000000011112222'
stud1.name = 'xiao ming'
stud1.fun()
stud1.hello('yang')

teacher1 = Teacher('professor', '10000650000011112222', 'yang pin')
Teacher.sex = 'female'
teacher1.fun()
teacher1.hello('wang')
```
