```python
# 随机仓位比
import numpy as np

def random_weight(num=5,times=8):
    weight = np.random.uniform(0, 1, size=[times, num])
    for i in range(0, times):
        a_sum = sum(weight[i])
        for j in range(0,num):
            weight[i][j]=round(float(weight[i][j]/a_sum),4) 
    return(weight)

# print(random_size(num=2,times=8))

# 基于仓位比计算 组合收益率
def portfolio_return(weight, meanList, num=5, times=8):
    portfolio = np.random.uniform(0,0,size =times).tolist()
    for i in range(0,times):
        for j in range(0,num):
            portfolio[i] = portfolio[i] + weight[i][j] * meanList[j]
    return portfolio


# 基于仓位比计算 组合标准差
def portfolio_std(weight, covList, times=8):
    portfolio = np.random.uniform(0,0,size =times).tolist()
    for i in range(0,times):
        var = np.dot(weight[i],covList)
        # print(var)
        var = np.dot(var,weight[i].T)  # 组合方差
        # print(var)
        portfolio[i] = np.sqrt(var)
    return portfolio

# 计算各组合的 夏普比
def sharp_ratio(rList, stdList, rf):
    sharpList = [(x[0]-rf)/x[1] for x in zip(rList,stdList)]
    return sharpList

```

```python
import pandas as pd
import numpy as np
import math
# 读取制定股票池
dfCode = pd.read_csv('./codeFile.csv',index_col = 0)
# 指定时间段 
start = '20200101'
end = '20201201'
# start = '20211001'
# end = '20211015'
# 一年的交易日天数
year = 252
#  散点个数
N = 2000


# print(dfCode)
print(dfCode.shape)
# 去除为nan的行
dfCode = dfCode.fillna(-1)
dfCode = dfCode[dfCode.code!=-1].reset_index(drop=True)

# 提取code列
dfCode = dfCode.loc[:,['code']]

# 去除重复的行
dfCode = dfCode.drop_duplicates().reset_index(drop=True)
# print(dfCode)
print(dfCode.shape)

# 提取为List
codeList = dfCode.code.values.tolist()
# print(codeList[:5])
codeList = [x[:-3] for x in codeList]
# print(codeList[:5])



# 提取收盘价
# 确定所需字段
fieldStr = ','.join(['secID','tradeDate','preClosePrice','closePrice'])
tickerList = codeList
# tickerList = codeList[:5]

dfData = DataAPI.MktEqudAdjGet(ticker=tickerList,beginDate=start,endDate=end,field=u"ticker,tradeDate,preClosePrice,closePrice,",pandas="1")

# print(dfData)
print(dfData.shape)
# 计算对数收益率
dfData['lnR'] = dfData.apply(lambda x: np.log(x.closePrice/x.preClosePrice),axis=1)

# 整理数据 
dfData = dfData.loc[:,['ticker','tradeDate','lnR']]

# 用groupby进行分类
dfList = list(dfData.groupby("ticker"))
# print(dfList[0][1])
# 准备第一个df用于循环拼接
dfNeed = dfList[0][1].loc[:,['tradeDate','lnR']].rename(columns={'lnR':dfList[0][0]})

# 循环拼接
for i in range(1,len(dfList)):
    dfNeed = pd.merge(dfNeed,
                      dfList[i][1].loc[:,['tradeDate','lnR']].rename(columns={'lnR':dfList[i][0]}),
                      how='left',
                      on='tradeDate')
# print(dfNeed)
dfNeed.index = dfNeed.loc[:,'tradeDate']

dfNeed = dfNeed.loc[:,tickerList]
# 剔除数据不全的标的
dfNeed = dfNeed.dropna(axis=1, how='all')
# print(dfNeed)


# 年化对数收益率
meanList = [x*year for x in dfNeed.mean().values]
# 年化对数收益率协方差矩阵
covList = (dfNeed.cov()*252).values.tolist()
# print(covList)

# 生成仓位比
weight = random_weight(num=dfNeed.shape[1], times=N )
# print(meanList)

# 组合收益率
portfolioReturn = portfolio_return(weight=weight, meanList=meanList, num=dfNeed.shape[1], times=N)

# print(portfolioReturn)

# 组合标准差
portfolioStd = portfolio_std(weight=weight, covList=covList, times=N)

# print(portfolioStd)



# 无风险利率
r = 0.019

portfolioSharp = sharp_ratio(rList=portfolioReturn, stdList=portfolioStd, rf=r)
# print(portfolioSharp)

# 各组合中 最小标准差
portfolioMinIndex = portfolioStd.index(min(portfolioStd)) 
# 各组合中 最大收益率
portfolioMaxIndex = portfolioReturn.index(max(portfolioReturn)) 
# 各组合中 最大夏普比
portfolioMaxIndex2 = portfolioSharp.index(max(portfolioSharp)) 



print('最小标准差:')
print(weight[portfolioMinIndex])
print(portfolioReturn[portfolioMinIndex])
print(portfolioStd[portfolioMinIndex])
print(portfolioSharp[portfolioMinIndex])
print('最大收益率:')
print(weight[portfolioMaxIndex])
print(portfolioReturn[portfolioMaxIndex])
print(portfolioStd[portfolioMaxIndex])
print(portfolioSharp[portfolioMaxIndex])
print('最大夏普比:')
print(weight[portfolioMaxIndex2])
print(portfolioReturn[portfolioMaxIndex2])
print(portfolioStd[portfolioMaxIndex2])
print(portfolioSharp[portfolioMaxIndex2])

# 绘图
import matplotlib.pyplot as plt

#用于正常显示中文标签
plt.rcParams["font.sans-serif"]=["SimHei"]
plt.rcParams["axes.unicode_minus"] = False

# 设置坐标轴的lable
plt.xlabel(" std ")
plt.ylabel(" return ")

colors1 = "blue"    #组合点的颜色
colors2 = "red"    #最大收益点
colors3 = "green"    #最小方差点
colors4 = "pink"    #最大sharpRatio
s = np.pi * 5 ** 2   # 普通组合点的大小
s1 = np.pi * 15 ** 2    # 最大瘦一点、最小方差点大小

# plt.title("马科维茨有效前沿")   
plt.scatter(portfolioStd, portfolioReturn, c=colors1, alpha=0.2,s=s)
plt.scatter(portfolioStd[portfolioMinIndex],portfolioReturn[portfolioMinIndex],c=colors2,alpha=0.8, s=s1, label = "max_retuen")
plt.scatter(portfolioStd[portfolioMaxIndex],portfolioReturn[portfolioMaxIndex],c=colors3,alpha=0.8,s=s1, label = "min_std")
plt.scatter(portfolioStd[portfolioMaxIndex2],portfolioReturn[portfolioMaxIndex2],c=colors4,alpha = 0.8,s=s1, label = "max_sharp")

plt.legend(loc = "best")
plt.plot()
plt.show()




```
