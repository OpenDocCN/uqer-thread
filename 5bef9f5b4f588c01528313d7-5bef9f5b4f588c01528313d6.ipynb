# 期权隐含波动率曲面计算

```python
from CAL.PyCAL import *
import pandas as pd
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import numpy as np


dd=DataAPI.OptDpoGet(secID=u"",optID=u"",ticker=u"",varSecID=u"",varTicker=u"510050",varType=u"",beginDate=u"20181115",endDate=u"20181115",exerType=u"",contractType=u"",field=u"",pandas="1")

data = dd[['tradeDate','optID','secShortName','ticker','strikePrice','listDate','expDate','preClosePrice','contractType','varName','varpreClosePrice','marginUnit']]
data['iv']=None

#print dd['ticker']
#tt=pd.DataFrame(columns=["opt_ticker", "strike_price", "exp_date","implied_volatility"])
exp_date=[]
imp_v=[]
strike_p=[]

for i in data['ticker']:
    t=1

    vol_data = data[data.ticker == i][data.tradeDate == '2018-11-15']
    types = vol_data.contractType.values[0]
    if types == 'CO' :
        opt_types = 1
    else :
        opt_types = -1
    strike = vol_data.strikePrice.values[0]
    spot = vol_data.varpreClosePrice.values[0]
    
    exp_date_str = vol_data.expDate.values[0]     # 期权行权日期
    eval_date_str = vol_data.tradeDate.values[0]   # 期权交易日期
    mat = Date.parseISO(exp_date_str) 
    evd = Date.parseISO(eval_date_str) 
    time_to_maturity = float(mat - evd + 1.0)/365.0 
    
    settle = vol_data.preClosePrice.values[0]
    
    calculated_vol = BSMImpliedVolatity(opt_types, strike, spot, 0.03, 0.01, time_to_maturity, settle)      # 计算隐含波动率
    # 返回的是dataframe格式，要转化为数值型
    calculated_vol = calculated_vol.iloc[0,0]
    # 期权的时间价值为负时(此种情况在50ETF期权里时有发生)，没法通过BSM模型计算隐含波动率，故此时将期权隐含波动率设为0.0，实际上，此时的隐含波动率和各风险指标并无实际参考价值
     # 但是为了防止分母为0，故设置为0.01%
    if calculated_vol is None:
        calculated_vol = 0.0001
    data.loc[data.ticker==i,'iv']= calculated_vol 
    #data[data['ticker']==i]['iv']=calculated_vol
    if types == 'CO' :
        exp_date.insert(t, time_to_maturity)
        
        imp_v.insert(t, calculated_vol)
        strike_p.insert(t, strike)
    else :
        pass
    t=t+1
    

```

```python
from CAL.PyCAL import *
import pandas as pd
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import numpy as np


fig = plt.figure()
ax = fig.gca(projection='3d')
 
# Make data.
x=np.array(exp_date,dtype=float)
y=np.array(strike_p,dtype=float)
#x, y = np.meshgrid(x, y)
z=np.array(imp_v,dtype=float)
print(x)
print(y)
print(z)
# Plot the surface.
surf = ax.plot_surface(x, y, z, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)
ax.set_xlabel('exp_date')
ax.set_ylabel('strike_p')
ax.set_zlabel('imp_v') 
ax.scatter(x, y, z)
plt.show()
```
