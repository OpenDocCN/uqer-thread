## 1、Alpha对冲模型简介

##### A、假设市场完全有效， 那么根据CAPM模型有， $R_{s} = R_{f}+\beta_{s}*(R_{m}-R_{f})$。其中 $R_{s}$表示股票收益，$R_{f}$表示无风险收益率，$R_{m}表示市场收益$，$\beta_{s}$表示股票相比于市场的波动程度，用以衡量股票的系统性风险。

##### B、遗憾的是， 市场并非完全有效， 个股仍然存在alpha（超额收益）。 根据Jensen's alpha 的定义：$\alpha_{s}=R_{s}-[R_{f}+\beta_{s}*(R_{m}-R_{f})]$，除掉被市场解释的部分，超越市场基准的收益即为个股alpha。

##### C、实际上， 股票的收益是受到很多方面因素的影响的， 比如说经典的Fama French 三因素就告诉我们， 市值大小、估值水平、以及市场因子就能解释股票收益， 而且低市值、低估值能够获取超额收益。那么、我们就可以通过寻找能够获取alpha的驱动因子来构建组合。

##### D、假设我们已经知道了哪些因子能够获取超额收益， 那么我们根据这些银子构建股票组合（比如持有低市值、低估值的股票）。那么组合的收益理论上是能够获取超额收益的，简单的来说， 组合的累计收益图应该是在基准（比如说沪深300）累计收益图之上的， 而且两者的差应该是扩大的趋势。

##### E、由于组合的涨跌我们是不知道的， 我们能够确保的是组合和基准的收益差在不断扩大， 那么持有组合， 做空基准， 对冲获得稳定的差额收益（alpha收益）， 这就是传说中的市场中性策略

## 2、优矿“三剑客”

针对上述研究流程，优矿提供了全程服务， 从金融大数据， 模型的研究到实盘交易和组合管理：

- DataAPI：提供近300个高质量的因子数据（基本面因子,技术面因子和大数据因子），为模型提供充足的原材料和让用户自己研究因子提供了基础
- RDP：提供标准的因子到信号的处理函数（`去极值、中性化、标准化`）同时，还提供了功能强大的组合构建函数
- Quartz：提供标准的、更贴近实际的回测框架，一键查看对冲模型历史表现

### 大赛的一些规则设计：

- #### 单子股票持股不超过10%：alpha收益并不是全仓某一只股票，然后涨停；可以看到，由于有股指期货的对冲，那么组合的持仓也应该像股指期货一样各行业配置很均匀，组合构建函数simple_long_only()充分考虑了行业配置、各因子alpha贡献等因素

- #### 股票仓位在任意三个以上收盘日低于80%则不达标：因为alpha收益已经非常稳健，那么增加本金的投入只会带来更多的收益，何乐而不为呢？

- #### 相对HS300，强制平仓线90%：近似于强平线为0.9，当所选因子不能持续带来alpha收益时，有必要对因子要仔细考虑了。

- #### 不能投资流动性过差以及刚上市的股票：alpha收益追求的是稳定性，没有必要去承受额外的流动性风险以及其他风险。

**在投资研究寻找新的因子就是专业量化研究员日常的工作，在研究过程中希望找到：有故事的因子，符合经济学原理，有投资逻辑**

学到的知识点：

1. alpha策略构建的步骤，首先获取初始因子值，然后对它们进行“去极值、中性化、标准化”的因子到信号的处理，最后将信号代入回测框架进行回测。这里组合构建使用的是RDP里面的simple_long_only()
2. 获取交易日列表的方法除了CAL.PyCAL里面的Date之外，还有uqer的接口TradeCalGet，它还可以获取是否是monthlastday之类的数据
3. 获取交易日前几日的交易日使用CAL.PyCAL里面的Date.advanceDate
4. 获取因子值的接口MktStockFactorsOneDayGet，需要注意的是下一步对因子值进行信号处理的三个处理函数的输入值格式为Dict/Series， key为ticker/secID, value为因子值，因此这里使用ticker设为index同时使用to_dict（）转化为dictionary
5. 将因子处理成信号之后，需要构建score矩阵，首先需要设置每个因子的权重，创建一个空的Dataframe，index为ticker，列为各个因子，然后根据权重计算score，最后将score列转化为dictionary
6. 最后将得到的score的dict代入到simple_long_only中，获得最后数据
7. 最后进行买卖。
