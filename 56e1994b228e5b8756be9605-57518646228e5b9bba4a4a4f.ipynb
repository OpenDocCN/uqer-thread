```python
import pandas as pd
import numpy as np
from datetime import datetime,timedelta
from CAL.PyCAL import *
cal = Calendar('China.SSE')
```

```python
#获得全A股的净利润
final_info_dic = {'ticker':[],'name':[],'NIncome':[]}
universe = DataAPI.EquGet(equTypeCD='A',listStatusCD=u"L",)['ticker'].tolist()   #获得全A股的ticker
for party_id in universe:
         #其他业绩快报等数据都有缺失，感谢小太阳提醒合并利润表数据比较全，会计期间是14年的
         s=DataAPI.FdmtISAllLatestGet(ticker=party_id,secID=u"",reportType=u"A",endDate=u"20141231",beginDate=u"20140101",field=u"secShortName,NIncome",pandas="1")
         if s['NIncome'].values!='':
            final_info_dic['ticker'].append(party_id)
            final_info_dic['name'].append(s['secShortName'].values)
            final_info_dic['NIncome'].append(s['NIncome'].iloc[0])
final_info_df = pd.DataFrame(final_info_dic)
```

```python
#展示
All_info_df_sort = final_info_df.sort(columns='NIncome',ascending=False).reset_index(drop=True)
All_info_df_sort['NIncome'] = np.round(All_info_df_sort['NIncome']/1e4,2).astype(str)+'万'
All_info_df_sort.columns = ['净利润','股票名称','股票编码']
print '股票利润排行'
All_info_df_sort
```

```python
All_info_df_sort.to_csv('option_data.csv')#导出CSV
```

```python
all=len(final_info_df.index)
print '上市公司总数：'
print all
print '上市公司净利润小于1500万的公司数：'
profit=len(final_info_df[final_info_df['NIncome']>150000000].index)
print profit
print '上市公司净利润亏损数：'
loss=len(final_info_df[final_info_df['NIncome']<0].index)
print loss
ii=float(profit)/float(all)
print '上市公司一年净利润深圳买不起房占比：'
print np.round(ii*100,2).astype(str)+'%'
```
