#  三均线BOLL共振的海龟交易策略（高抛低吸法加趋势跟踪加仓法）

三均线BOLL共振的海龟交易策略（高抛低吸法加趋势跟踪加仓法）

海龟交易系统本质上是一个趋势跟随的系统,但是最值得我们学习的是资金管理尤其是分批建仓及动态止损的部分。现在中国的行情以震荡市为主，独立的海龟交易策略侧重建仓条件为上穿BOLL上轨或唐奇安通道上轨，清仓条件为下穿BOLL下轨或唐奇安通道下轨，这种趋势跟踪的策略不适合现阶段的中国行情，A股行情的震荡市更适合高抛低吸的原则，思路可以跟传统海龟交易原则的突破想法相反，通过超买超卖判断建仓、离场时间点。也就是股价上穿BOLL下轨代表股票超卖可以建仓，下穿BOLL上轨代表股票超买可以离场。当然海龟的经典ATR加仓、减仓的仓位控制我们仍然保留，再加上长中短移动平均线、MACD中的DIF和DEA、VOL等条件共振去判断建仓、离场点。这种方法既能考虑震荡市去高抛低吸，也能跟踪上趋势来加仓减仓。总体比较下来，此策略能取得跑赢沪深300从2015年1月1日到2018年的基准行情。

1、建仓选择
（1）某股票现价上穿20日BOLL下轨
（2）MA5金叉MA10，且MA5、MA10、MA20多头向上排列
（3）VOL(5,10)，MAVOL5上穿MAVOL10
（4）MACD(12,26,9),DIF>DEA & DIF>0 & DEA>0
（5）某股票不在仓位中
满足条件（1）或（2-5)全满足可以建仓，每个股票建仓仓位控制在10%，如果一次有多只股票满足，满仓后不再选择新的股票建仓。


2、加仓条件
预备：
N值是仓位管理的核心，涉及加仓及止损。另外，N值与技术指标平均真实波幅ATR很相似，用ATR代替.
首先介绍真实波幅： 真实波幅是以下三个值中的最大值
（1）当前交易日最高价和最低价的波幅
（2）前一交易日的收盘价与当前交易日最高价的波幅
（3）前一交易日的收盘价与当前交易日最低价的波幅
公式： 
TrueRange=Max(High−Low,abs(High−PreClose),abs(PreClose−Low))
接下来，N值计算公式为：
N=(PreN[−19：]+TrueRange)/20
其中 preN为前面N值，TrueRange为当前的真实波幅,此公式的含义为计算之前20天（包括今天在内）的N的平均值
另外，有些海龟交易系统用的是ATR来代替N值，ATR为真实波幅的20日平均。

最终条件：若股价在上一次买入（或加仓）的基础上上涨了0.5N，则加仓一个Unit。最多加仓4个unit
接上面的例子：假如N值仍为0.2。
价格来到 5 + 0.2*0.5 = 5.1时，加仓1个Unit，买入5000股，耗资25500元，剩余资金 49500元
价格来到 5.1 + 0.2*0.5 = 5.2 时再加仓1个unit。买入5000股，耗资26000元，剩余资金 23500元


3、止盈或止损（清仓）
（1）某股票现价下穿20日BOLL上轨
（2）MA5下穿MA10
（3）量均线VOL5下穿VOL10
（4）相比最后一次买入价下跌了2个ATR
（1）-（4）只要有1个条件满足清仓。
