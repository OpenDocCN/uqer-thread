```python
# 引入工具包
from __future__ import division 
from lib.lib import *
import scipy.stats as st
sns.set_style('white')

from quartz.api import *
```

```python
# 定义因子分析类 factorAnalyseMachine
class factorAnalyseMachine(object):
    
    def __init__(self, factor, date = '', period = 10, universe = 'A'):
        '''date用于确认给出基本信息的时间，period是计算一段时间的RankIC时所用间隔，universe在给出基本信息和计算timeRangeRankIC时用到'''
        self.cal = Calendar('China.SSE')
        self.factor = factor
        self.date = self.cal.advanceDate(dt.datetime.today(), '-2B').strftime('%Y%m%d') if date == '' else date
        self.period = period
        self.universe = universe
        self.factorBenchmark = ['PE', 'ROA', 'LCAP', 'ARTRate', 'AR', 'EPS', 'CTOP', 'QuickRatio', 'EGRO', 'REC', 'NetProfitGrowRate']
        
        self.industryDist = pd.DataFrame()
        self.ICDataFrame = pd.DataFrame()
        self.direction = None
    
#********************************************
# 初识因子
#********************************************
    def _corrDisplayFactorBenchmark(self, corrData, figsize = None):
        '''利用相关系数、p值构成的DataFrame画出条形图，x轴是因子'''
        fig = plt.figure(figsize = figsize)
        ax1 = fig.add_subplot(111)
        corrData.sort('r', ascending = False, inplace = True)
        corrData['xtick'] = range(len(corrData))
        sigPos = corrData[(corrData.r > 0)&(corrData.p < 0.01)]
        sigNeg = corrData[(corrData.r < 0)&(corrData.p < 0.01)]
        nonSig = corrData[~(corrData.p < 0.01)]
        ax1.bar(sigPos.xtick, sigPos.r, align = 'center', color = '#F08080', label = u'显著为正')
        ax1.bar(nonSig.xtick, nonSig.r, align = 'center', color = 'black', label = u'不显著')
        ax1.bar(sigNeg.xtick, sigNeg.r, align = 'center', color = 'g', label = u'显著为负')
        # plt.legend([u'显著为正', u'不显著', u'显著为负'], loc = 1, prop = legendFont)
        plt.xticks(corrData.xtick, [factorNameDict[x] for x in corrData.index], fontproperties = font)
        plt.xlim(-1, len(corrData))
        fig.autofmt_xdate(rotation = 30, ha = 'right')
        plt.grid(axis = 'y')

    def describeFactor(self, winsor = True, neutral = True, standard = True):
        '''考察universe内股票在tradeDate这天的factor值分布情况，以及factor与基准因子风格的相关性情况'''
        factor = self.factor; universe = self.universe; tradeDate = self.date; 
        Univ = set_universe(universe, tradeDate) if type(universe)==str else universe

        
        print '本次关注的因子是：'
        print factorDataFrame[factorDataFrame[u'名称'] == factor][u'描述'].values[0]
        print '='*100,u'\n','在',universe,'中，',factorNameDict[factor],'(%s)'%factor,'的分布情况是：'
        field = (','.join(self.factorBenchmark) + ',ticker,' if 
                 factor in self.factorBenchmark else ','.join(self.factorBenchmark) + ',ticker,' + factor)
        df = DataAPI.MktStockFactorsOneDayGet(tradeDate = tradeDate, secID = Univ, 
                field = field).set_index('ticker')
        
        # winsorize
        if winsor:
            for x in [factor]+self.factorBenchmark:
                df[x] = pd.Series(winsorize(df[x].to_dict(), win_type = 'QuantileDraw', pvalue = 0.05))
        if neutral:
            for x in [factor]+self.factorBenchmark:
                df[x] = pd.Series(neutralize(df[x].to_dict(), self.date, 'short', 'SW1'))
        if standard:
            for x in [factor]+self.factorBenchmark:
                df[x] = pd.Series(standardize(df[x].to_dict()))
                
        df[factor].hist(bins = 20, normed = True, color = '#B2DFEE')
        plt.grid(axis = 'x')
        plt.show()


        corrDict = {}
        for x in self.factorBenchmark:
            if x == factor:
                continue
            corrDf = df[[factor, x]].dropna(axis = 'index')
            corrDict[x] = st.pearsonr(corrDf[factor], corrDf[x])
        pearsonDf = pd.DataFrame(corrDict, index = ['r','p']).T
        print '='*100,u'\n','在',universe,'中，',factorNameDict[factor],'(%s)'%factor,'与基准因子池的相关性为：'
        
        self._corrDisplayFactorBenchmark(pearsonDf)
        return pearsonDf
    
    
    def factorIndustryDist(self, winsor = True, neutral = True, standard = True):
        '''给出因子在指定日期时在各个行业的分布情况，返回列名为Q20,Q50,Q80的DataFrame'''
        factor = self.factor; tradeDate = self.date
        
        res = {}
        for x in industryDict:
            tmpSeries = DataAPI.MktStockFactorsOneDayGet(tradeDate, 
                secID = set_universe(industryDict[x], tradeDate), field = 'ticker,'+factor).set_index('ticker')[factor]
            if len(tmpSeries)==0:
                res[x] = [np.nan]*3
                continue
                
                
            if winsor:
                tmpSeries = pd.Series(winsorize(tmpSeries.to_dict(),win_type = 'QuantileDraw', pvalue = 0.05))
            if neutral:
                tmpSeries = pd.Series(neutralize(tmpSeries.to_dict(),self.date, 'short', 'SW1'))  
            if standard:
                tmpSeries = pd.Series(standardize(tmpSeries.to_dict()))
            
            
            tmpSeries.dropna(inplace = True)
            res[x] = [tmpSeries.quantile(0.2), tmpSeries.median(), tmpSeries.quantile(0.8)]
            self.industryDist = pd.DataFrame(res, index = ['Q20', 'Q50', 'Q80']).T
        return pd.DataFrame(res, index = ['Q20', 'Q50', 'Q80']).T


    def displayIndustryDist(self, winsor = True, neutral = True, standard = True):
        '''对于factorIndustryDist()给出的结果进行图形化，默认因子名称是self.factor'''
        if self.industryDist.empty:
            self.factorIndustryDist(winsor, neutral, standard)
        distDataFrame = self.industryDist
            
            
        fig = plt.figure(figsize = (16,6))
        ax1 = fig.add_subplot(111)
        distDataFrame.sort_values('Q50', ascending = False).Q80.plot(kind = 'bar', ax = ax1, color = '#D8BFD8')
        distDataFrame.sort_values('Q50', ascending = False).Q50.plot(kind = 'bar', ax = ax1, color = '#DB7093')
        distDataFrame.sort_values('Q50', ascending = False).Q20.plot(kind = 'bar', ax = ax1, color = '#F08080')

        plt.legend([u'80%分位点', u'中位数', u'20%分位点'], prop = font)
        plt.xticks(fontproperties = font, rotation = -90)
        plt.title(u'%s在各行业中分布情况'%self.factor, fontproperties = titleFont)
        plt.grid(axis = 'y') 
        
        
#*****************************************
# IC / Rank IC - 关注盈利预测能力
#*****************************************

    def ICCalc(self, tradeDate = '', 
               samplePeriod = 10, predictPeriod = 10, winsor = True, rank = True):
        '''计算给定因子股票池时，tradeDate前samplePeriod天因子均值与之后predictPeriod天前瞻收益率的相关系数与p值
        
            如果rank为True，返回秩相关系数，否则相关系数
        '''
        if tradeDate == '':
            tradeDate = self.cal.advanceDate(dt.datetime.today(), '-'+str(predictPeriod+2)+'B').strftime('%Y%m%d')
        factor = self.factor; universe = self.universe
        
        Univ = set_universe(universe, tradeDate) if type(universe)==str else universe

        beginPrice = DataAPI.MktEqudAdjGet(tradeDate = tradeDate, secID = Univ, 
                        field = 'tradeDate,closePrice,secID').pivot(
                        index = 'tradeDate', values = 'closePrice', columns = 'secID')
        beginPrice.index = ['periodReturn']
        endPrice = DataAPI.MktEqudAdjGet(tradeDate = self.cal.advanceDate(tradeDate,'+'+str(predictPeriod)+'B'), secID = Univ, 
                        field = 'tradeDate,closePrice,secID').pivot(
                        index = 'tradeDate', values = 'closePrice', columns = 'secID')
        endPrice.index = ['periodReturn']
        data = (endPrice/beginPrice - 1).T


        factorAvg = {}
        for sec in Univ:
            factorAvg[sec] = DataAPI.MktStockFactorsDateRangeGet(secID = sec, field = factor, 
                            beginDate = self.cal.advanceDate(tradeDate,'-'+str(samplePeriod - 1)+'B'), 
                            endDate = tradeDate).mean(axis = 'index').values[0]   
        data[factor] = pd.Series(factorAvg)
        data.dropna(axis = 'index', inplace = True)

        data.index = map(lambda x: x[:6], data.index)
        
        if winsor:
            for x in data.columns:
                data[x] = pd.Series(winsorize(data[x].to_dict(), win_type = 'QuantileDraw', pvalue = 0.05))
            

        return st.pearsonr(data.periodReturn.rank(), data[factor].rank()) if rank else st.pearsonr(data.periodReturn, data[factor])


    def displayCorrTime(self, figsize = (16,6)):
        '''利用相关系数、p值构成的DataFrame画出条形图，x轴为时间'''
        
        if self.ICDataFrame.empty:
            self.timeRangeRankIC()
        corrData = self.ICDataFrame
        
        fig = plt.figure(figsize = figsize)
        ax1 = fig.add_subplot(111)
        # corrData.sort('r', ascending = False, inplace = True)
        corrData['xtick'] = range(len(corrData))
        sigPos = corrData[(corrData.r > 0)&(corrData.p < 0.01)]
        sigNeg = corrData[(corrData.r < 0)&(corrData.p < 0.01)]
        nonSig = corrData[~(corrData.p < 0.01)]
        ax1.bar(sigPos.xtick, sigPos.r, align = 'center', color = '#F08080', label = u'显著为正')
        ax1.bar(nonSig.xtick, nonSig.r, align = 'center', color = 'black', label = u'不显著')
        ax1.bar(sigNeg.xtick, sigNeg.r, align = 'center', color = 'g', label = u'显著为负')
        plt.legend(loc = 'best', prop = legendFont)
        plt.xticks(corrData.xtick, corrData.index, fontproperties = font)
        plt.xlim(-1, len(corrData))
        fig.autofmt_xdate(rotation = -90, ha = 'center')
        plt.grid(axis = 'y')
        plt.title(u'因子Rank IC值', fontproperties = titleFont)


    def timeRangeRankIC(self, beginDate = '20160101', endDate = '', period = ''):
        '''返回一段时间内因子的rank IC值，将samplePeriod和predictPeriod都设为period'''
        
        if period == '':
            period = self.period
        if endDate == '':
            endDate = self.cal.advanceDate(dt.datetime.today(), '-'+str(period+2)+'B').strftime('%Y%m%d')
        factor = self.factor; universe = self.universe
        Univ = set_universe(universe, beginDate) if type(universe)==str else universe
        factorIC = {}


        for date in [x[1] for x in timeSplit(beginDate, endDate, period)]:
            factorIC[date] = self.ICCalc(tradeDate = date, samplePeriod = period, predictPeriod = period, winsor = True, rank = True)
            print date,
        print 'Finished.'
        self.ICDataFrame = pd.DataFrame(factorIC, index = ['r', 'p']).T
        self.direction = len(self.ICDataFrame[self.ICDataFrame > 0]) / len(self.ICDataFrame) > 0.5
        return pd.DataFrame(factorIC, index = ['r', 'p']).T

    
#************************************
# 回测表现
#************************************
    def QuantileBt(self, beginQuantile = 0, endQuantile = 0.2, dropNegtive = False, 
                        start = '2014-01-01', end = '2016-08-13', benchmark = 'HS300', 
                        universe = set_universe('A', '2014-01-01'), capital_base = 3000000, freq = 'd', refresh_rate = 20, 
                        winsor = True, neutral = True, standard = True):
        '''在universe中筛选给定因子值在给定分位范围内的股票进行回测'''
        def initialize(account):
            pass
        
        def handle_data(account):                  # 每个交易日的买入卖出指令
            quantileSec = factorQuantileFilter(self.factor, 
                                               account.universe, account.previous_date, beginQuantile, endQuantile, dropNegtive,
                                              winsor = winsor, neutral = neutral, standard = standard)

            buylist = [x for x in quantileSec.index if isST(x, account.previous_date)==False]
            buylist = [x for x in buylist if (x in account.referencePrice and account.referencePrice[x]>0)]
            for stk in account.avail_secpos:
                order_to(stk, 0)
            for stk in buylist:
                order_pct_to(stk, 1/len(buylist))
            return
        
        sim_params = qz.SimulationParameters(start, end, benchmark, universe, capital_base)
        #sim_params = qz.sim_condition.env.SimulationParameters(start, end, benchmark, universe, capital_base)
        idxmap, data = qz.get_backtest_data(sim_params)
        #idxmap, data = qz.sim_condition.data_generator.get_daily_data(sim_params)
        strategy = qz.TradingStrategy(initialize, handle_data)
        #strategy = qz.sim_condition.strategy.TradingStrategy(initialize, handle_data)
        
        bt, account = qz.quick_backtest(sim_params, strategy, idxmap, data, refresh_rate = refresh_rate)
        perf = qz.perf_parse(bt, account)
        
        return bt,perf
    
    def multiQuantileBt(self, group = 5, dropNegtive = False, 
                        start = '2014-01-01', end = '2016-08-13', benchmark = 'HS300', 
                        universe = set_universe('A', '2014-01-01'), capital_base = 3000000, freq = 'd', refresh_rate = 20,
                       winsor = True, neutral = True, standard = True):
        '''返回多组回测结果，分组直接以universe中股票因子值为基准'''
        quans = map(lambda x: round(x, 2), list(np.arange(0,1,1/group)) + [1.0])
        quansPerf = {}
        quansBt = {}
        for x in range(group):
            tmpBt, tmpPerf = self.QuantileBt(quans[x], quans[x+1], dropNegtive, start, end, benchmark, 
                                         universe, capital_base, freq, refresh_rate,
                                            winsor = winsor, neutral = neutral, standard = standard)
            quansPerf[str(x)] = tmpPerf
            quansBt[str(x)] = tmpBt
            print x,
        return quansBt, quansPerf
    
    
    def industryQuantileBt(self, beginQuantile = 0, endQuantile = 0.2, dropNegtive = False, 
                        start = '2014-01-01', end = '2016-08-13', benchmark = 'HS300', 
                        universe = set_universe('A', '2014-01-01'), capital_base = 3000000, freq = 'd', refresh_rate = 20,
                          winsor = True, neutral = True, standard = True):
        '''在universe中筛选不同行业内给定因子值在给定分位范围内的股票进行回测'''
        def initialize(account):
            pass
        
        def handle_data(account):                  # 每个交易日的买入卖出指令
            
            buylist = []
            for x in industryDict:              
                quantileSec = factorQuantileFilter(self.factor, set_universe(industryDict[x], account.previous_date),
                                               account.previous_date, beginQuantile, endQuantile, dropNegtive,
                                                  winsor = winsor, neutral = neutral, standard = standard)
                buylist += list(quantileSec.index)
            buylist = [x for x in quantileSec.index if isST(x, account.previous_date)==False and x in account.universe]
            buylist = [x for x in buylist if (x in account.referencePrice and account.referencePrice[x]>0)]
            for stk in account.avail_secpos:
                order_to(stk, 0)
            for stk in buylist:
                order_pct_to(stk, 1/len(buylist))
            return
        
        sim_params = qz.SimulationParameters(start, end, benchmark, universe, capital_base)
        #sim_params = qz.sim_condition.env.SimulationParameters(start, end, benchmark, universe, capital_base)
        idxmap, data = qz.get_backtest_data(sim_params)
        #idxmap, data = qz.sim_condition.data_generator.get_daily_data(sim_params)
        strategy = qz.TradingStrategy(initialize, handle_data)
        #strategy = qz.sim_condition.strategy.TradingStrategy(initialize, handle_data)
        bt, account = qz.quick_backtest(sim_params, strategy, idxmap, data, refresh_rate = refresh_rate)
        perf = qz.perf_parse(bt, account)
        
        return bt,perf
    
    
    def industryMultiQuantileBt(self, group = 5, dropNegtive = False, 
                        start = '2014-01-01', end = '2016-08-13', benchmark = 'HS300', 
                        universe = set_universe('A', '2014-01-01'), capital_base = 3000000, freq = 'd', refresh_rate = 20,
                        winsor = True, neutral = True, standard = True):
        '''返回多组回测结果，在各个行业中选取相应分位数分组'''
        quans = map(lambda x: round(x, 2), list(np.arange(0,1,1/group)) + [1.0])
        quansPerf = {}
        quansBt = {}
        for x in range(group):
            tmpBt, tmpPerf = self.industryQuantileBt(quans[x], quans[x+1], dropNegtive, start, end, benchmark, 
                                         universe, capital_base, freq, refresh_rate,
                                        winsor = winsor, neutral = neutral, standard = standard)
            quansPerf[str(x)] = tmpPerf
            quansBt[str(x)] = tmpBt
            print x,
        return quansBt, quansPerf
```
