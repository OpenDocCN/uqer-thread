## 1.1
介绍之前3因子模型
\begin{align*}
R_i(t)–R_F(t) = a_i + b_i(R_M(t) – R_F(t))+ s_iSMB_t + h_iHML_t + e_i
\end{align*}
说明三因子模型中，还有未能解释的部分。引出5因子模型
\begin{align*}
R_i(t) – R_F(t) = a_i + b_i(R_M(t) – R_F(t)) + s_iSMB_t + h_iHML_t + r_iRMW_t + c_iCMA_t + e_i
\end{align*}

# 二.我的理解和因子构造
## 2.1
\begin{align*}
R_i(t)–R_F(t)
\end{align*}作为期望超额收益，就是把这个期望超额收益放入一个回归方程，用因子来解释这个超额收益是怎么回事。
三因子模型就是找到了三个因子来很好的解释这个超额收益。
那么，如果三因子模型中的三个因子可以完全解释各种风险带来的超额收益，任何一个股票以及任何一个投资组合的$a_i$的真实值应该为0。但是，很多组合是显著不为0，这说明三因子模型中的变量并不能解释所有超额收益。后来，就有了五因子模型。
增加了两个因子之后，五因子模型就更好的解释了超额收益。

###### 这东西能赚钱吗？
>两个个简单策略
如果说，五因子模型能够完美的解释投资组合，那么，任何一个股票以及任何一个投资组合的$a_i$的真实值应该为0。我们相信$a_i$为0，那么，买入在这一期$a_i$值为负的组合，静静等候$a_i$到0的时候就可以了。
如果说，五因子模型可以解释大部分，却不能完美解释投资组合，那么，我们就相信$a_i$大于0是显著的。对与$a_i$大于0的投资组合就是能获利的组合。

## 2.2

### 市场因子
\begin{align*}
R_M(t) – R_F(t)
\end{align*}
CAMP模型中的因子
>一个对camp因子的策略
$b_i$的大小之间表示了相对于大盘收益风险，$b_i$越大表示了其期望利润越高。

### 市值因子
市值因子是指公司的规模对该公司股票的风险有着接影响：资产规模小，风险就会相对增加，收益也会增加，反之，资产规模大，风险就会相对减少，收益减少。
将市场中公司股票按市值排列。前一半为大市值股票，后一半为小市值股票。那么，第一个因子$SMB$为
\begin{align*}
E(SMB)=E(r_S)-E(r_B)
\end{align*}
其中，$r_S$为小市值平均期望收益，$r_B$为大市值股票

### 市账因子
市账因子就是账面的所有者权益除以市值（下简称B/M）。市账因子风险描述了公司的额外财务困境风险，说明市场上对公司的估值比公司自己的估值要低。这些公司一般都是销售状况或者盈利能力不是十分好的公司，因此相对于低B/M的公司来说需要更高的收益来补偿。
将市场中公司股票按市账因子排序。选前30%与后30%分别作为高，低市帐公司。
\begin{align*}
E(HML)=E(r_H)-E(r_L)
\end{align*}
其中，$r_H$为高市帐公司平均收益率，$r_L$为低市帐公司平均收益率。

以上为三因子模型中三个因子

### 盈利因子与投资因子
盈利水平风险是指，盈利能力较高的行业一般会伴随着更高的风险。记做 $E(RMW)$，其计算方法和$E(HML)$类似（也是将股票分成三份，然后计算高/低盈利水平的股票期望收益率之差）
投资水平可以用再投资率来衡量，我们认为投资率偏低的公司风险较大，投资者对这些公司有更高的收益率要求，反之亦然。计算方法与$E(HML)$类似。

以上，五因子齐了

```python
import numpy as np
import pandas as pd 
from CAL.PyCAL import *
import statsmodels as sm
import statsmodels.api as sm
import matplotlib.pyplot as plt
from statsmodels.sandbox.regression.predstd import wls_prediction_std
```

```python
Year = 2012
data = DataAPI.TradeCalGet(exchangeCD=u"XSHG",beginDate=str(Year*10000+501),endDate=str((Year+2)*10000+601),field=['calendarDate','isMonthEnd'])
data = data[data['isMonthEnd'] == 1]
date= map(lambda x: x[0:4]+x[5:7]+x[8:10], data['calendarDate'].values.tolist())
mounthdate = list(data.calendarDate)
```

```python
#求某月的因子值
def Areturn(i = 0):
    breakpoint = date[i]                      
    #取中证800
    universe = set_universe('000906.ZICN',date=breakpoint)                          
    ##SMB股票选取市场规模因子
    ME = DataAPI.MktEqudGet(tradeDate=breakpoint,secID=universe,field=u"ticker,marketValue").dropna()   
    ME30 = np.percentile(ME['marketValue'],50) 
    ME70 = np.percentile(ME['marketValue'],50)
    SM = ME[ME['marketValue']<=ME30]['ticker'].tolist()                               
    BM = ME[ME['marketValue']>ME70]['ticker'].tolist()
    ##HMI股票选取市账率因子
    BP = DataAPI.MktStockFactorsOneDayGet(tradeDate=breakpoint,secID=universe,field=u"ticker,PB").dropna() 
    BP = BP[BP>0].dropna()                                                  
    BP[['PB']]=1/BP[['PB']]                                                
    BP30 = np.percentile(BP['PB'],30)
    BP70 = np.percentile(BP['PB'],70)
    LP = BP[BP['PB']<=BP30]['ticker'].tolist()                                      
    HP = BP[BP['PB']>BP70]['ticker'].tolist()
    ##RMW股票选取盈利因子
    PR = DataAPI.MktStockFactorsOneDayGet(tradeDate=breakpoint,secID=universe,field=u"ticker,OperatingProfitRatio").dropna() 
    PR30 = np.percentile(PR['OperatingProfitRatio'],30) 
    PR70 = np.percentile(PR['OperatingProfitRatio'],70)
    LR = PR[PR['OperatingProfitRatio']<=PR30]['ticker'].tolist()                               
    HR = PR[PR['OperatingProfitRatio']>PR70]['ticker'].tolist()
    ##CMA股票选取投资因子
    CA = DataAPI.MktStockFactorsOneDayGet(tradeDate=breakpoint,secID=universe,field=u"ticker,TotalAssetGrowRate").dropna() 
    CA30 = np.percentile(CA['TotalAssetGrowRate'],30) 
    CA70 = np.percentile(CA['TotalAssetGrowRate'],70)
    LC = CA[CA['TotalAssetGrowRate']<=CA30]['ticker'].tolist()                                
    HC = CA[CA['TotalAssetGrowRate']>CA70]['ticker'].tolist()


    tickerlist = [SM,BM,LP,HP,LR,HR,LC,HC]
    tickername = ['SM','BM','LP','HP','LR','HR','LC','HC']
    #取出选择股票计算收益率
    cfrm = pd.DataFrame()
    for n,j in zip(tickername,tickerlist):
        a = DataAPI.MktEqudGet(tradeDate=date[i],secID=u"",ticker= j ,beginDate=u"",endDate=u"",isOpen="",field=u"ticker,	tradeDate,accumAdjFactor,closePrice",pandas="1")
        a['price'] = a['closePrice'] * a['accumAdjFactor']
        b = DataAPI.MktEqudGet(tradeDate=date[i+1],secID=u"",ticker= j ,beginDate=u"",endDate=u"",isOpen="",field=u"ticker,	tradeDate,accumAdjFactor,closePrice",pandas="1")
        b['price'] = b['closePrice'] * b['accumAdjFactor']
        c = ((b['price'] - a['price'])/a['price']).mean()
        cfrm[n] = [c]
    return cfrm
```

```python
Returnfrm = pd.DataFrame()
k = 0
for t in range(len(mounthdate)-1):
    x = Areturn(t)
    Returnfrm = pd.concat([Returnfrm,x])
    k = k+1
    print k,
Returnfrm.index = mounthdate[1:]
```
