# 分享一个反转策略，回馈uqer

刚刚接触python，刚刚接触uqer。感觉还挺不错的。这里有很多量化投资的干货，不是哪一个网站都免费帮你梳理的。

从零看了一天python教程，就开始实践，这个周末各种debug。这里向广大和我一样的新手推荐一个debug的好东西，就是print！回测的时候多print几个变量，你就知道自己哪里出问题了。

回归正题，这个反转策略的想法受羊驼策略启发，比较简单：

1. 每个回测周期在股票池中选取过去超跌的股票list（参数自己定），完成初次建仓
2. 初次建仓后，每个回测周期，先卖出一只持仓中过去表现最差的stock，再买入一只整个股票池里表现最差的stock
3. 依次循环

看到回测的结果，仔细想想其实还是有些逻辑的。HS300这个白马众多的股票池，超跌后左侧买入是较好的建仓位置。同时，应华尔街名言“截断亏损，让利润奔跑”，每个交易周期卖出让你损失最多的股票是理想的平仓策略。
