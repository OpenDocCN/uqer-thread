```python
from heapq import nlargest, nsmallest
import numpy as np
# 计算相关性矩阵行情起止时间
start = '20110101'
end = '20130101'
# 股票池，如果是 n 只则相关系数矩阵含有 n**2 个元素，有效相关系数有 n*(n-1)/2 个
# ！！如果设成全A股需要运行很久
stocklist = set_universe('HS300', start)
# 获取行情
stlen = len(DataAPI.MktIdxdGet(indexID = '000300.ZICN', beginDate = start, endDate = end, field='closeIndex')) # 标准长度
allclsprx = []
for i,s in enumerate(stocklist):
    z = DataAPI.MktEqudAdjGet(secID = s, beginDate = start, endDate = end, field='closePrice')
    if len(z) != stlen:
        stocklist.remove(s)
    else:
        allclsprx.append(z['closePrice'].tolist())
idxmap = {i:s for i,s in enumerate(stocklist)}
allclsprx = np.matrix(allclsprx)
# 计算相关系数矩阵，获取相关系数列表，排序
covmat = np.corrcoef(allclsprx, y=None, rowvar=1, bias=0, ddof=None)
d = len(covmat)
covlist = []
for i in range(d-1):
    for j in range(i+1, d):
        covlist.append((covmat[i, j], i, j))
# 挑出负相关系数与正相关系数最大的各10只股票
neglist = nsmallest(10, covlist, key=lambda x:x[0])
neglist = [map(idxmap.get, p[1:])+[p[0]] for p in neglist]
poslist = nlargest(10, covlist, key=lambda x:x[0])
poslist = [map(idxmap.get, p[1:])+[p[0]] for p in poslist]
```

下面是自己写的回测测试，思路是找出上轨以上的位置编号sell_l 下轨以下的位置编号buy_l  在均值以上0.3倍std和均值以下2std作为sell的平仓范围ping_sell
均值以下0.3倍std和均值以上2倍std作为buy的平仓范围ping_buy   根据这些进出场位置算收益

```python
#获得了两个 序列 delta1 delta2   
#感觉 回到均值附近就可以 平仓了 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt  
c=2.1 #c*std
d=3
c1=2
d1=0.3
c2=0.3
d2=2
window_long=30
sell_l=[]
buy_l=[]
ping_buy=[]
ping_sell=[]
for i in range(window_long,len(delta1)-1):
    mean=np.mean(delta1[i-window_long:i])
    std=np.std(delta1[i-window_long:i])
    strike_price_buy=mean - c*std
    strike_price_sell=mean + d*std
    if delta1[i+1]<=strike_price_buy:
        buy_l.append(i+1)
    if delta1[i+1]>=strike_price_sell:
        sell_l.append(i+1)
    if delta1[i+1]>=mean-c1*std and delta1[i+1]<=mean+d1*std:
        ping_sell.append(i+1)
    if delta1[i+1]>=mean-c2*std and delta1[i+1]<=mean+d2*std:
        ping_buy.append(i+1)
print buy_l
print sell_l
print ping_buy
print ping_sell
```

上面是进出场的位置（delta序列中的位置）

buy_l 和 ping_buy 配对 

```python
sum_t=0
sum_l=[0]
j_pos=-1
for i in range(0,len(ping_buy)):
    for j in range(j_pos+1,len(buy_l)):
        if buy_l[j]<ping_buy[i]:
            sum_t+=delta1[ping_buy[i]]-delta1[buy_l[j]]
            sum_l.append(sum_t)
            j_pos=j
        else:
            break
print sum_l    
xlist=sum_l
ylist=range(0,len(sum_l))
plt.plot(ylist,xlist)
```

sell_l 和 ping_sell 的配对

```python
sum_t2=0
sum_l2=[0]
j_pos2=-1
for i in range(0,len(ping_sell)):
    for j in range(j_pos2+1,len(sell_l)):
        if sell_l[j]<ping_sell[i]:
            sum_t2+=delta1[sell_l[j]]-delta1[ping_sell[i]]
            sum_l2.append(sum_t2)
            j_pos2=j
        else:
            break
print sum_l2   
xlist=sum_l2
ylist=range(0,len(sum_l2))
plt.plot(ylist,xlist)
```

把两者的收益结合一起

```python
sum_tot=0
sum_tot_l=[]
a_sign=min(len(sum_l),len(sum_l2))
a1=sum_l[0:a_sign]
a2=sum_l2[0:a_sign] 
for i in range(0,max(len(sum_l),len(sum_l2))):
    if i<a_sign:
        sum_tot=a1[i]+a2[i]
        sum_tot_l.append(sum_tot)
    else:
        sum_tot+=(sum_l2[i]-sum_l2[i-1])
        sum_tot_l.append(sum_tot)
print sum_tot_l
xlist=sum_tot_l
ylist=range(0,len(sum_tot_l))
plt.plot(ylist,xlist)
```
