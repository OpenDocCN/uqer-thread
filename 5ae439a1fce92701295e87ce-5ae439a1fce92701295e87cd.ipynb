# **模型训练**
-----
为了能让模型及时抓取到市场的变化，我们采用了滚动回测方法。模型训练区间为20070101至20171231，按月份划分（前12个月训练，本月测试），因此需要不断重复训练。
get_train_val_test_data()：数据集划分函数
format_feature_label()：特征划分函数
pipeline():模型处理，包括因子选择，格点搜索等等

利用格点搜索做交叉验证，选择最好的模型，在测试集测试

注意：
	部分函数的编写参考Uqer社区的 华泰证券 《人工智能选股之 Boosting 模型》这篇文章

```python
def get_train_val_test_data(date):
        #计算前12个月，采用365天
        d_12=(datetime.datetime.strptime(date, "%Y-%m") - datetime.timedelta(days=365)).strftime('%Y-%m')
        train_df=df[(df['y_m']>=d_12) & (df['y_m']<date)]
        #train_val_df = train_val_df.sample(frac=1).reset_index(drop=True)   
        #拆分训练集、验证集                                 
        test_df =df[df['y_m']==date]   
        return train_df,test_df

def format_feature_label(origin_df, is_filter=True):
    
    #if is_filter:
        #origin_df = origin_df[origin_df['rets']!=0]
    df_test= origin_df                
    feature = origin_df[factor_names]
    label = origin_df['rets']

    return feature, label,df_test

```

```python
#def pipeline():
LR_model_list = []
accuracy=[]
classification=[]
final=pd.DataFrame()

for date in df['y_m'].drop_duplicates()[12:]:
    

    print('training model for %s' % date)
    train_df,test_df = get_train_val_test_data(date)
    train_feature, train_label,df_train = format_feature_label(train_df)       
    test_feature, test_label,df_test = format_feature_label(test_df)
    params1 = {'C':range(1,10,2),'gamma':[x /100. for x in range(1,10,2)],'kernel':['linear','rbf','sigmoid']}
    clf_svm = svm.SVC(probability=True)
   #   

    clf_AdaBoost = AdaBoostClassifier(n_estimators = 350 , learning_rate = 1.53 ) 
    
    lr=LogisticRegression()           
    params={'C':np.arange(0.01,0.1,0.01)}
    #clf=GridSearchCV(lr,param_grid=params,)
    ada=AdaBoostClassifier( ) 
    params3={'n_estimators':[200,400,600,800,1000],'learning_rate':[0.5,1,1.5]}
    #clf=GridSearchCV(ada,param_grid=params3)
    clf= GridSearchCV(clf_svm, params1)
    clf.fit(train_feature,train_label)
    clf_best=clf.best_estimator_
    #clf_best=svm.SVC(C=9,gamma=0.09,kernel='rbf',probability=True)
    #clf_best=AdaBoostClassifier(n_estimators = 800 , learning_rate = 1.53 ) 
    print(clf_best)
    

    predict_prob=clf_best.predict_proba(test_feature)[:,1] 


    accuracy.append(clf_best.score(test_feature,test_label))
    LR_model_list.append(clf_best)
    df_test.loc[:,'factor'] = predict_prob

    lables=df_test['factor']>df_test['factor'].mean()
    lables = lables.replace({True:'1',False:'0'})
   # lables.columns=['lables']
   # df_test=pd.concat([df_test,lables],axis=1)
    df_test.loc[:,'lables'] = lables
    #df_test.loc[:,'rets'] = predict_value
    df_test = df_test.loc[:, ['stock', 'tradeDate', 'rets', 'factor','lables']]
    classification.append(metrics.classification_report(df_test['rets'],df_test['lables']))
    final=pd.concat([final,df_test])
    #return LR_model_list,final,accuracy
```

# **因子回测**
上述模型在测试的时候，为每个股票打了一个分数，可以认为是将输入的因子转换成了一个集成的“因子”，之后我们选取排名前30%的的因子进行选股。
考虑到Uqer回测和读取数据的速度问题，这里仅仅选取HS300的股票回测，读者自行可以选择ZZ500，或者全A股

```python
#导入上述模型计算结果
signal_df =final
factor_data = signal_df[['stock', 'tradeDate', 'factor']]     # 读取因子数据
factor_data = factor_data.set_index('tradeDate', drop=True)
q_dates = factor_data.index.values
```

```python


# 可编辑部分与 strategy 模式一样，其余部分按本例代码编写即可

# -----------回测参数部分开始，可编辑------------
start = '2014-04-01'                       # 回测起始时间
end = '2018-03-01'                         # 回测结束时间
benchmark = 'HS300'  
universe = DynamicUniverse('HS300')
capital_base=10000000
freq = 'd'                                      # 'd'表示使用日频率回测，'m'表示使用分钟频率回测
refresh_rate = Monthly(1)
# 执行handle_data的时间间隔

# ---------------回测参数部分结束----------------
# 把回测参数封装到 SimulationParameters 中，供 quick_backtest 使用
sim_params = quartz.SimulationParameters(start, end, benchmark, universe, capital_base)
# 获取回测行情数据
idxmap, data = quartz.get_daily_data(sim_params)
# 运行结果
results = {}
# 调整参数(选取股票的Q因子五分位数)，进行快速回测
for quantile_five in [1,0.5]:
    
    # ---------------策略逻辑部分----------------
    refresh_rate = Monthly(1)  
    commission = Commission(0.0003, 0.0013)
   
    
    def initialize(account):                   # 初始化虚拟账户状态
        pass
    
    def handle_data(account):                  # 核心策略逻辑    
        #account = context.get_account('fantasy_account')
        #current_universe = account.universe('stock', exclude_halt=True)
        current_universe = account.universe
        pre_date = account.previous_date.strftime("%Y-%m-%d")
        if pre_date in q_dates:                          
            # 拿取调仓日前一个交易日的因子，并按照相应十分位选择股票
            q = factor_data.ix[pre_date].dropna()       
            q = q.set_index('stock', drop=True)
            q = q.ix[current_universe].dropna() 
            q_max = q['factor'].quantile(quantile_five)
            q_min= q['factor'].quantile(quantile_five-0.5)
            my_univ = q[(q['factor']>=q_min)&(q['factor']<q_max)].index.values 
            #print(context.current_date)
            #print(my_univ)
            # 交易部分
            positions = account.security_position
            #print(positions)
            sell_list = [stk for stk in positions if stk not in my_univ]
            for stk in sell_list:   
                account.order_to(stk,0)     
            #在目标股票池中的，等权买入
            for stk in my_univ:
                account.order_pct_to(stk, 1.0/len(my_univ))
# ---------------策略逻辑部分结束----------------

    # 把回测逻辑封装到 TradingStrategy 中，供 quick_backtest 使用
    strategy = quartz.TradingStrategy(initialize, handle_data)
    # 回测部分
    bt, acct = quartz.quick_backtest(sim_params, strategy, idxmap, data, refresh_rate=refresh_rate)

    # 对于回测的结果，可以通过 perf_parse 函数计算风险指标
    perf = quartz.perf_parse(bt, acct)

    # 保存运行结果
    tmp = {}
    tmp['bt'] = bt
    tmp['annualized_return'] = perf['annualized_return']
    tmp['volatility'] = perf['volatility']
    tmp['max_drawdown'] = perf['max_drawdown']
    tmp['alpha'] = perf['alpha']
    tmp['beta'] = perf['beta']
    tmp['sharpe'] = perf['sharpe']
    tmp['information_ratio'] = perf['information_ratio']
    
    results[quantile_five] = tmp
    print (str(quantile_five))
print( 'done'   )
```

```python
import matplotlib.pyplot as plt
from matplotlib import rc
rc('mathtext', default='regular')
import seaborn as sns
sns.set_style('white')
from matplotlib import dates
import time
import scipy.stats as st
from CAL.PyCAL import font    # CAL.PyCAL中包含font


fig = plt.figure(figsize=(14,9))
fig.set_tight_layout(True)
ax1 = fig.add_subplot(111)
#ax2 = fig.add_subplot(212)
ax1.grid()
#ax2.grid()

for qt in results:
    bt = results[qt]['bt']

    data = bt[[u'tradeDate',u'portfolio_value',u'benchmark_return']]
    data['portfolio_return'] = data.portfolio_value/data.portfolio_value.shift(1) - 1.0   # 总头寸每日回报率
   # data['portfolio_return'].ix[0] = data['portfolio_value'].ix[0]/	1e8 - 1.0
    data['excess_return'] = data.portfolio_return - data.benchmark_return                 # 总头寸每日超额回报率
    data['excess'] = data.excess_return + 1.0
    data['excess'] = data.excess.cumprod()                # 总头寸对冲指数后的净值序列
    data['portfolio'] = data.portfolio_return + 1.0     
    data['portfolio'] = data.portfolio.cumprod()          # 总头寸不对冲时的净值序列
    data['benchmark'] = data.benchmark_return + 1.0
    data['benchmark'] = data.benchmark.cumprod()          # benchmark的净值序列
    results[qt]['hedged_max_drawdown'] = max([1 - v/max(1, max(data['excess'][:i+1])) for i,v in enumerate(data['excess'])])  # 对冲后净值最大回撤
    results[qt]['hedged_volatility'] = np.std(data['excess_return'])*np.sqrt(252)
    results[qt]['hedged_annualized_return'] = (data['excess'].values[-1])**(252.0/len(data['excess'])) - 1.0
    #data[['portfolio','benchmark','excess']].plot(figsize=(12,8))
   # ax.plot(data[['portfolio','benchmark','excess']], label=str(qt))
    ax1.plot(data['tradeDate'], data[['portfolio']], label=str(qt))
    
    #ax2.plot(data['tradeDate'], data[['excess']], label=str(qt))
ax1.plot(data['tradeDate'], data[['benchmark']], label=u'HS300')    

ax1.legend([u'Bottom 50% stocks',u'Top 50% stocks',u'HS300'],loc=0)
#ax2.legend(loc=0)
ax1.set_ylabel(u"净值", fontproperties=font, fontsize=16)
#ax2.set_ylabel(u"对冲净值", fontproperties=font, fontsize=16)
```
