	公司当前净债务7.59亿，假设未来三年企业经营活动现金流为14.5亿、15.76亿、17.18亿；未来三年投资现金10.2亿、11.24亿、12亿；永续增长率为4%，平均资本成本为10%，请回答下列问题：

    3.1 计算当前公司企业价值

    3.2 计算当前公司股权价值


作业1

```python
bs = {}
bs['总资产'] = 4
bs['股东权益'] = 2.5
print('负债：%.2f亿' % (bs['总资产'] - bs['股东权益']))

r = {}
r['净利润'] = 0.3
r['总费用'] = 1.75
print('营收：%.2f亿' % (r['净利润'] + r['总费用']))

e = {}
e['股东权益'] = 2.3
e['净支付'] = 0.1
e['回购'] = 0
e['增发'] = 0
print('分红：%.2f亿' % (e['净支付'] - e['回购'] - e['增发']))
```

作业2

```python
a = {}
a['存款'] = 100
a['利率'] = 0.05
a['分红'] = 5
a['收入'] = a['存款'] * a['利率']

print('----资产负债表----')
a['负债'] = 0
a['现金资产'] = a['存款'] + a['收入'] - a['分红']
print('现金资产：%.2f万' % a['现金资产'])
print('股东权益：%.2f万' % (a['现金资产'] - a['负债']))

print('\n----利润表----')
a['成本'] = 0
print('成本：%.2f万' % a['成本'])
print('利润：%.2f万' % (a['收入'] - a['成本']))

print('\n----现金流量表----')
a['投资活动净额'] = 0
print('经营活动净额：%.2f万' % a['收入'])
print('投资活动净额：%.2f万' % a['投资活动净额'])
print('融资活动净额：%.2f万' % -a['分红'])
```

```python
f = {}
f['永续增长率'] = 0.04
f['平均资本成本'] = 0.1
f['净债务'] = 7.59

df = pd.DataFrame()
df['经营现金流'] = [14.5, 15.76, 17.18]
df['投资现金'] = [10.2, 11.24, 12]
df['自由现金流'] = df['经营现金流'] - df['投资现金']
df['折现率'] = [(1 + f['平均资本成本'])**(i+1) for i in range(len(df))]
df['自由现金流现值'] = df['自由现金流'] / df['折现率']
print(df.to_html())

f['永续现金流'] = df['自由现金流'].iloc[-1] * (1 + f['永续增长率']) / (f['平均资本成本'] - f['永续增长率'])
f['永续现金流现值'] = f['永续现金流'] / df['折现率'].iloc[-1]
f['企业价值'] = df['自由现金流现值'].sum() + f['永续现金流现值']
f['股权价值'] = f['企业价值'] - f['净债务']
print('企业价值：%.2f亿' % f['企业价值'])
print('股权价值：%.2f亿' % f['股权价值'])
```
