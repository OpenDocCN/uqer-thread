```python
# 【实盘感悟】遇上暴跌我该怎么做？ https://uqer.datayes.com/v3/community/share/565d47e3f9f06c6c8a91af49

"""
内容摘要：作者统计了上证综指在历史上的周度跌幅，发现周度跌幅都低于15%；随后统计了上证综指连续5天的跌幅，以及指数累计下跌10%、15%、20%之后，在随后几天指数的反弹情况，分别从反弹幅度和反弹概率上进行了统计；结果发现，在大盘累计下跌超过10%之后再抄底，在未来5天内都能实现正的盈利，而且下跌幅度越高之后再抄底，获得的收益越高。
"""
"""
记得在7,8月份大跌时，我做过一个简单的统计，统计了上证综指历史上所有周度的涨跌幅，统计后发现，大部分的涨跌幅都位于（-5%，5%），一些极端情况会出现涨跌10%左右，跌的最多的好像也就15%左右，而且即使在熊市，跌多了也会短期反弹一点，用专业的话讲，应该算是下跌趋势中的短期反转吧（momentum&amp;reverse）。
    所以，当时面对当时的下跌，我并没有太绝望，我在等的也就是反转的机会，而且不承担选股风险，就买带杠杆的分级B，快进快出，这样熊市下来我并没有亏损太多，甚至略微有盈利。
    那么问题来了，怎么判定反转呢？不然这些都是大白话。我的想法也非常简单，既然历史（当时主要是07,08年的）统计表明跌幅超过15%之后再继续下跌的概率不高，那么一旦跌幅超过15%我在入场抄底，反转的机会是不是很高呢？说起来很简单吧，但做起来呢？
    做起来就没那么简单了，人毕竟是感性的，遇上满仓跌停你的第一感觉就是将亏损和工资or奖金匹配了，这个时候都是恐惧，所以才出现了巨幅的跳水，因为大部分人都已经没法正常思考了（比如你当时想买某个分级B，你绝对不太会像之前一样理性，先要看看分级基金整体折溢价如何。。）
    但这个时候就是真正的机会，别人恐惧时你要贪婪。想在想想当时7、8月份的行情，有很多曾经盘中一度跌幅超过15%的吧，这其实都是很好的抄底时机，但在熊市里，抄底要注意快进快出！
"""
"""
所谓分级B是指一类具有杠杆属性，与其它杠杆投资方式相比不涉及信用账户、保证金等问题，投资者通过券商交易系统像买卖股票一样进行买卖，购买方式灵活，场内买入投资起点低，交易便利的分级基金。
"""

"""
本例已和原例无太大关系了；
"""
```

```python
start_date = '20180801' # 开始日期
end_date = (dt.datetime.today() - dt.timedelta(days=1)).strftime('%Y%m%d') # 截止日期

# 交易所交易日历 
"""
XSHG，XSHE，CCFX，XDCE，XSGE，XZCE，XHKG。XSHG表示上海证券交易所，XSHE表示深圳证券交易所，CCFX表示中国金融期货交易所，XDCE表示大连商品交易所，XSGE表示上海期货交易所，XZCE表示郑州商品交易所，XHKG表示香港证券交易所。可同时输入多个证券交易所,可以是列表
"""
TradeCal=DataAPI.TradeCalGet(exchangeCD="XSHG",beginDate=start_date,endDate=end_date,field="calendarDate,isOpen",pandas="1")
TradeCal = TradeCal.loc[TradeCal['isOpen'] == 1,:] # TradeCal = TradeCal[TradeCal['isOpen'] == 1]

Dates = map(lambda x: x[0:4]+x[5:7]+x[8:10], TradeCal['calendarDate'])
data = pd.DataFrame(index=Dates, data=0, columns=['total PB(LCAP)','total PE(LCAP)','total PB(LFLO)','total PE(LFLO)','shangzheng','shenzheng','chuangye','hs300','shzh50','zhzh500','LCAP','LFLO'])

for date in Dates:
    """
    上证50(SH50)、上证180(SH180)、沪深300(HS300)、中证500(ZZ500)、创业板(CYB)、中小板(ZXB)、全A股(A)，如"HS300"；另外，支持自定义     指数代码，可以获得date日期的成分股列表，如"000001.ZICN"；对于预设行业实例（IndSW/IndZJH/IndZZ）及预设指数实例（IdxCN），可直接使     用date (datetime.datetime): 交易日
    """
    universe = set_universe('A', date) # + set_universe('ZZ500', date)
    
    # 获取多只股票历史上某一天的因子数据
    pe_pb = DataAPI.MktStockFactorsOneDayGet(tradeDate=date,secID=universe[:],field=['ticker','LCAP','LFLO','PB','PE'],pandas="1").set_index('ticker')
    """
    LCAP	float	对数市值（Natural logarithm of total market values）。计算方法：对数市值=市值的对数。属于价值类因子。数据更新     时间：T日21点30分。
    LFLO	float	对数流通市值（Natural logarithm of float market values）。计算方法：对数流通市值=流通市值的对数。属于价值类因     子。数据更新时间：T日21点30分。
    """
    data_pe_pb = float(pe_pb.shape[0])
    pe_pb.dropna(inplace=True)
    # 数据量不足80%时，不具有代表性
    if float(pe_pb.shape[0]) / data_pe_pb < 0.8:
        print('Error.')
        continue
    
     # 细节处理：负值极值处理
    pe_pb['PB'][pe_pb['PB'] < 0] = 0
    pe_pb['PE'][pe_pb['PE'] < 0] = 0
    pb = winsorize(pe_pb['PB'].to_dict())
    pe = winsorize(pe_pb['PE'].to_dict())
    pe_pb['PB'][pb.keys()] = pb.values()
    pe_pb['PE'][pe.keys()] = pe.values()
    
    # copy记录 LCAP LFLO
    pe_pb['LCAP_sum'] = pe_pb['LCAP']
    pe_pb['LFLO_sum'] = pe_pb['LFLO']
    
    # 权重化
    pe_pb['LCAP'] = pe_pb['LCAP'] / sum(pe_pb['LCAP'])
    pe_pb['LFLO'] = pe_pb['LFLO'] / sum(pe_pb['LFLO'])
    
    # 获取指数日线行情信息，包含昨收价、开盘价、最高价、最低价、收盘价、成交量、成交金额等字段。
    MktIdxdGet_1=DataAPI.MktIdxdGet(ticker=['000001','399001','399006','000300','000016','000905'],tradeDate=date,field='closeIndex',pandas="1")
    
    data.loc[date] = [sum(pe_pb['LCAP'] * pe_pb['PB']), sum(pe_pb['LCAP'] * pe_pb['PE']),sum(pe_pb['LFLO'] * pe_pb['PB']), sum(pe_pb['LFLO'] * pe_pb['PE']),MktIdxdGet_1.iloc[0,0],MktIdxdGet_1.iloc[1,0],MktIdxdGet_1.iloc[2,0],MktIdxdGet_1.iloc[3,0],MktIdxdGet_1.iloc[4,0],MktIdxdGet_1.iloc[5,0],sum(pe_pb['LCAP_sum']),sum(pe_pb['LFLO_sum'])]

# 权重化
data['LCAP'] = data['LCAP'] / sum(data['LCAP'])
data['LFLO'] = data['LFLO'] / sum(data['LFLO'])

# 增加4列
data['average PB(LCAP)'] = data['total PB(LCAP)']*data['LCAP']
data['average PE(LCAP)'] = data['total PE(LCAP)']*data['LCAP']
data['average PB(LFLO)'] = data['total PB(LFLO)']*data['LFLO']
data['average PE(LFLO)'] = data['total PE(LFLO)']*data['LFLO']
data['average PB(LCAP)'] = sum(data['average PB(LCAP)'])
data['average PE(LCAP)'] = sum(data['average PE(LCAP)'])
data['average PB(LFLO)'] = sum(data['average PB(LFLO)'])
data['average PE(LFLO)'] = sum(data['average PE(LFLO)'])

data_Relative = pd.DataFrame(index=Dates, data=0, columns=['shangzheng','shenzheng','chuangye','hs300','shzh50','zhzh500'])

data.tail()
```
