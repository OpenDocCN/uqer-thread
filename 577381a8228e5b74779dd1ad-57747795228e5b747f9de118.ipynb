```python
# 日期格式处理函数
def itoDateStr(year = 2016, month = 6, day = 26):
    return str(year) + ('0'+str(month) if (month<10) else str(month))+('0'+str(day) if (day<10) else str(day))


def toDateTime(argStr = u"2012-01-01"):
    return datetime.date(string.atoi(argStr[0:4]), string.atoi(argStr[5:7]), string.atoi(argStr[8:10]))


# 获取连续期货价格函数
def getFutuCon(bDate = u"20120101", eDate = u"", conObj = u"AU", period = 1):
    histPrice = {}
    if eDate == "":
        eDate = time.strftime("%Y%m%d",  time.localtime(time.time()))
        
    calendar = DataAPI.TradeCalGet(exchangeCD = u"XSHE", beginDate = bDate, endDate = eDate, field = u"calendarDate,isOpen", pandas = "1")
    calArray = calendar[calendar['isOpen'] == 1]['calendarDate'].values
    
    for date in calArray:
        year = string.atoi(date[2:4]);month = string.atoi(date[5:7])
        year = (year if (month+period<= 12) else year+1)
        month = (month+period if (month+period<= 12) else month+period-12)
        year = str(year)
        month = ('0'+str(month) if(month<10) else str(month))
        tker = conObj+year+month
        priceVal = DataAPI.MktFutdGet(tradeDate = toDateTime(date), ticker = tker, field = u"closePrice")
        histPrice[toDateTime(date)] = (np.NaN if len(priceVal) == 0 else priceVal)
        
    return pd.DataFrame(histPrice, index = [conObj+'_Continuous_'+str(period)]).T
```
