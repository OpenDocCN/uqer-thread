# 数据获取以及转存

1. 获取数据
--------------------
***

我们提供了``DataAPI``帮助你获取各种行情，基本面数据。

```python
data = DataAPI.MktEqudAdjGet(ticker = '600000')
```

```python
data.tail()
```

2. 存储数据
------------------
***

数据保存只需要一行代码：（可以储存为``.csv``或者``.xlsx``格式，分别通过``to_csv``和``to_excel``方法）

```python
data.to_csv(u'spd_data.csv')
```

接着就可以在你的``data``页下看到你保存的数据。

![image](http://ww2.sinaimg.cn/large/63ef6a8dgw1es8b86zblyj21g30oswpx.jpg)

3. 数据下载
-------------------
***

所有在``Data``页面下的数据都可以通过点击下载图标下载至本地：

4. 数据上传
-------------------------
***

数据上传可以通过``Data``页面下的``上传``按钮完成：

![](http://ww1.sinaimg.cn/large/63ef6a8dgw1esjqtvtyurj20cc0a63zw.jpg)

上传完成后的数据会在``Data``页面下显示，然后在我们的环境中当作本地文件读入，例如：

```python
import pandas as pd
data = pd.read_excel(u'alpha基金数据.xlsx')
```
