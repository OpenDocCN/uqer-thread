# 营业收入增长率 + 市盈率  from deligne

## 策略思路
----
- 股票池设为营业收入增长率最大的300只股票与市盈率最低的300只股票的交集
- 以上股票池中，选择营业收入增长率最大的20只股票，平均买入
- 一个月调仓一次

### 本策略得到的超出基准HS300指数的收益曲线图

```python
bt['alpha'] = ((bt['portfolio_value']/bt['portfolio_value'][0] - 1) - ((1 + bt['benchmark_return']).cumprod() - 1))
bt = bt.set_index('tradeDate')
bt['alpha'].plot(figsize=(14,7))
```
