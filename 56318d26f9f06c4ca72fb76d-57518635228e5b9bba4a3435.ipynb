# 【学酥小白Q&A】海通量化策略实现——波动率在趋势确认中的应用

### 策略思路：
- 当价格上涨，突破长期移动平均线一定比例时买入：Pt > [1+threshold]\* MA\_t(60)，其中threshold为过去60日对数价格的标准差std(lnPt(60))
- 当价格下跌，突破中期移动平均线时卖出：Pt < MA\_t(20)
