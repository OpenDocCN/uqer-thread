# 基金三因子研究

# 原理
基于FF三因子公式，假设基金收益可以用以下方程分解：
$r_p-r_f=\alpha+b_p(r_m-r_f)+s_pSMB_p+h_pHML_p+\epsilon_p$
使用基金在期间的收益率序列和各风格的收益率序列做多元回归，就可以近似求出基金在各种风格上的分布。
***
# 函数说明
## get_tradedate_before
这是我自己的函数库中的函数，输入结束日期和时间长度N，返回结束日期之前N个交易日列表。

## get_all_stock_fund
返回当前所有开放式混合基金

## my_ols
输入x和y（y可以是多列），返回回归结果

## get_fund_series(fund,start_date,end_date)
输入基金代码，起始日期结束日期，返回日期内收益率序列

## get_base_series(base,start_date,end_date)
输入基准指数（这里使用中证全指），返回日期内收益率序列

## get_ref_series(base,end_date,period)
**功能：计算风格收益率。**
输入基准指数，根据基准指数的成分股计算HML、SMB序列。注意HML是低估值收益率-高估值收益率，SMB序列是大市值收益率-小市值收益率。
返回基准指数收益率序列、SMB收益率序列、HML收益率序列

## cal_style(fund,rm,r_SMB,r_HML,end_date,period):
**主函数，计算基金风格**

***

# 使用方法：
如果要计算多个基金，由于市场收益率和风格收益率会多次用到，因此首先调用get_ref_series计算风格收益率，然后再调用主函数计算基金风格。

```python
import pandas as pd
import numpy as np
from scipy.optimize import minimize 
import statsmodels.api as sm
from lib.TradeDate import *
rf=0.00011
```

```python
# 获取所有场内混合基金
def get_all_stock_fund():
    funds=DataAPI.FundGet(listStatusCd=u"",category=u"H",operationMode=u"O",field=u"ticker",pandas="1")
    return list(funds['ticker'])
```

```python
#OLS回归
def my_ols(dataX,dataY):
    X=sm.add_constant(dataX)
    model = sm.OLS(dataY,X).fit()
    return model.params
```

```python
#计算基金收益率序列
def get_fund_series(fund,start_date,end_date):
    data=DataAPI.FundNavGet(ticker=fund,beginDate=start_date,endDate=end_date,field=u"endDate,adjNavChgPct",pandas="1")
    data.set_index('endDate',inplace=True)
    data=data/100
    return data
```

```python
# 获取基准收益率序列
def get_base_series(base,start_date,end_date):
    #计算市场收益
    dataBase=DataAPI.MktIdxdGet(ticker=base,field='tradeDate,CHGPct',beginDate=start_date,endDate=end_date,pandas='1')
    dataBase.set_index('tradeDate',inplace=True)
    dataBase['CHGPct']=dataBase['CHGPct']
    return dataBase['CHGPct']
```

```python
#计算RM、SMB、HML序列
def get_ref_series(base,end_date,period):
    start_date=get_tradedate_before(end_date,period)[0]
    #计算市场收益率
    rm=get_base_series(base,start_date,end_date)
    #使用中证全指成分股
    secID=list(DataAPI.IdxConsGet(secID=base,intoDate=start_date,field=u"consID",pandas="1")['consID'])
    #获得起始日期所有股票流通市值和估值数据
    data=DataAPI.MktEqudGet(secID=secID,tradeDate=start_date,field=u"secID,negMarketValue,PB",pandas="1")
    data.set_index('secID',inplace=True)
    data.dropna(inplace=True)
    data=data[data['PB']>0]
    #获取市值和估值前三分之一和后三分之一的代码
    S=list(data.sort_values(by='negMarketValue',ascending=True).iloc[:int(len(data)/3)].index)
    B=list(data.sort_values(by='negMarketValue',ascending=True).iloc[-int(len(data)/3):].index)
    H=list(data.sort_values(by='PB',ascending=True).iloc[-int(len(data)/3):].index)
    L=list(data.sort_values(by='PB',ascending=True).iloc[:int(len(data)/3)].index)
    #计算样本时间内各因子收益率
    chgPct=DataAPI.EquRetudGet(secID=list(data.index),beginDate=start_date,endDate=end_date,
                               isChgPctl=u"0",field=u"secID,tradeDate,dailyReturnReinv",pandas="1")
    chgPct=chgPct.groupby(['secID','tradeDate']).mean().unstack()
    pS=chgPct[chgPct.index.isin(S)].mean(axis=0).loc['dailyReturnReinv']
    pB=chgPct[chgPct.index.isin(B)].mean(axis=0).loc['dailyReturnReinv']
    pH=chgPct[chgPct.index.isin(H)].mean(axis=0).loc['dailyReturnReinv']
    pL=chgPct[chgPct.index.isin(L)].mean(axis=0).loc['dailyReturnReinv']
    r_SMB=pB-pS
    r_HML=pL-pH
    return rm,r_SMB,r_HML
```

```python
# 多元回归计算风格
def cal_style(fund,rm,r_SMB,r_HML,end_date,period):
    start_date=get_tradedate_before(end_date,period)[0]
    result={}
    #整合市场、SMB、HML序列
    df_style=pd.concat([rm,r_SMB,r_HML],axis=1)
    df_style.columns=['market','SMB','HML']
    #获取基金收益率序列
    r_fund=get_fund_series(fund,start_date,end_date)
    #回归分析
    res=my_ols(df_style,r_fund)
    result['alpha']=round(res['const']*100,2)
    result['RM']=round(res['market'],2)
    result['SMB']=round(res['SMB'],2)
    result['HML']=round(res['HML'],2)
    return result
```

```python
fund=get_all_stock_fund()
end_date='2021-09-15'
period=120
index_market='000985'#中证全指
result={}
rm,r_SMB,r_HML=get_ref_series(index_market,end_date,period=period)
for f in fund:
    try:
        r=cal_style(f,rm,r_SMB,r_HML,end_date,period=period)
        result[f]=r
    except:
        pass
```
