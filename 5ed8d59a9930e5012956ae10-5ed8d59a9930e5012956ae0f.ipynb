# Andy_计算因子的IC值

```python
# 计算20100101——20171231，每月末的ROE因子与下一个月收益率的相关系数
import numpy as np
import pandas as pd

#获取交易日历20100101——20171231，每月末的交易日期
calendar_frame = DataAPI.TradeCalGet(exchangeCD=u"XSHG,XSHE",beginDate=u"20100101",\
                                     endDate=u"20171231",isOpen=u"1",field=u"calendarDate,isOpen,isMonthEnd",pandas="1")
calendar_frame =calendar_frame[calendar_frame['isMonthEnd']==1]
calendar_frame['next_monthend'] = calendar_frame['calendarDate'].shift(-1)#正数为向下移动
calendar_frame=calendar_frame[['calendarDate','next_monthend']]
calendar_frame.rename(columns = {'calendarDate':'monthend'},inplace=True)#好
calendar_frame.loc[3887,'next_monthend']='2018-01-31'
calendar_frame.head()
```

```python
calendar_frame_list=calendar_frame['monthend'].tolist()
calendar_frame_list
```

```python
#获取沪深300的股票代码
tickers = DataAPI.IdxConsGet(secID=u"",ticker=u"000300",isNew=u"",intoDate=u"20141231",\
                             field=u"consTickerSymbol,consShortName",pandas="1")
tickers = tickers['consTickerSymbol'].tolist()
tickers
```

```python
#获取300只股票的每月的因子
factor_frame_list = []
for i in calendar_frame_list:
    factor_frame=DataAPI.MktStockFactorsOneDayGet(tradeDate=i,secID=u"",\
                                     ticker=tickers,field=u"ticker,tradeDate,ROE,DAVOL10",pandas="1")
    factor_frame_list.append(factor_frame)
tfactor_frame = pd.concat(factor_frame_list)#方法很好，得到多天多只股票得数据
#tfactor_frame['tradeDate']=factor_frame['tradeDate'].apply(lambda x:x.replace('-',''))
tfactor_frame.rename(columns = {'tradeDate':'monthend'},inplace = True)
tfactor_frame.head()
```

```python
#获取300只股票的下一个月的收益率
monthly_return_list=[]
for i in tickers:
    monthly_return_frame_i = DataAPI.MktEqumGet(ticker=i,\
                   beginDate=u"20100101",endDate=u"20180131",field=u"ticker,secShortName,endDate,return",pandas="1")
    monthly_return_frame_i['nextmonth_return']=monthly_return_frame_i['return'].shift(1)
    monthly_return_frame_i=monthly_return_frame_i.dropna()
    monthly_return_frame_i=monthly_return_frame_i.drop('return',axis = 1)
    
    monthly_return_list.append(monthly_return_frame_i)
    
monthly_return_frame=pd.concat(monthly_return_list)


monthly_return_frame.rename(columns = {'endDate':'monthend'},inplace = True)
monthly_return_frame.head()
```

```python
len(tfactor_frame)
```

```python
#将每一个月的因子和下一月的收益率组合在一起
monthly_return_sum=pd.merge(monthly_return_frame,tfactor_frame,on = ['ticker','monthend'],how ='left')
monthly_return_sum.head()
```

```python
#定义相关系数函数
import scipy.stats as st
def normal_ic1(df):
    tmp_df = df.copy()
    tmp_df.dropna()
    return tmp_df['nextmonth_return'].corr(tmp_df['ROE'])
def normal_ic2(df):
    tmp_df = df.copy()
    tmp_df.dropna()
    return st.pearsonr(tmp_df['nextmonth_return'],tmp_df['ROE'])[0]#返回两个值，第一个值为相关系数，第二个值为P值
def rank_ic(df):
    tmp_df = df.copy()
    tmp_df.dropna()
    return st.spearmanr(tmp_df['nextmonth_return'],tmp_df['ROE'])[0]
```

```python
#分别计算在不同函数下的相关系数
normal_ic1_frame1 = monthly_return_sum.groupby(['monthend']).apply(normal_ic1) #按照日期进行分组，计算每一组的先关系数
normal_ic1_frame1.plot()
```

```python
normal_ic1_frame2 = monthly_return_sum.groupby(['monthend']).apply(normal_ic2)
normal_ic1_frame2.plot()
```

```python
normal_ic1_frame3= monthly_return_sum.groupby(['monthend']).apply(rank_ic)
normal_ic1_frame3.plot()
```
