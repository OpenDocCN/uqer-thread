# 基于montecarlo模拟法计算VaR

```python
import pandas as pd
import numpy as np
```

在用montecarlo模拟法计算VaR，数据是18个组合的单位净值，想问为什么重复次数对VaR结果会有这么大影响？

```python
def VaR(df_concat):
    df=df_concat.dropna()
    df_seri=df.ix[:,0]
    me=df.ix[:,1].mean()/20
    st=df.ix[:,1].std()/(20**0.5)       
    nav_now=df_seri[-1]
    nav_sample=list()
    np.random.seed(0)
    for j in range(1000):
        s=np.random.randn(20)
        for t in range(20):
            nav_next=nav_now*(1+(me+st*s[t]))
            nav_now=nav_next
        nav_sample.append(nav_now)
    nav_sample.sort()
    a=df_seri[-1]
    b=nav_sample[int(len(nav_sample)*0.05)]
    c=a-b
    d=c/a
    return a,b,c,d
portdata=pd.read_excel('portdata20161031.xlsx',index_col=0)

for i in range(18):
    df=pd.concat([portdata.ix[:,i],portdata.pct_change().ix[:,i]],axis=1)
    print(i ,VaR(df))
```
