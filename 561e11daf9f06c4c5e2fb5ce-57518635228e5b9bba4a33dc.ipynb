# 用5日均线和10日均线进行判断---国科大量化投资

```python
data=DataAPI.MktEqudGet(ticker="600030",beginDate="20131001")   #选取600030股票
a=data.closePrice
B = []
n=len(a)
for i in range(10,n):
    x5=0
    for j in range(i-5,i):
        x5=x5+a[j]
    x5=x5/5              #5日均线值
    x10=0
    for k in range(i-10,i):
        x10=x10+a[k]
    x10=x10/10               #10日均线值
    if x5>x10:
        B.append(1)
    else:
        B.append(0)
```

```python
import matplotlib.pyplot as plt
o=data.openPrice
m= len(B)
w=0                 #利润
price0=1000000          #操作金额1亿，但考虑买的份额为100的整数，取1百万
Price0=[]             
Price=[]              #利润w的数组
for i in range(1,m):
    if i==m:
        w=0            #最后一天利润为0
    else:
        if B[i-1]==0 and B[i]==1:     
            k1=i+10            
            amount=price0//o[k1]     #买入份额
        elif B[i-1]==1 and B[i]==0:
            k2=i+10
            price0=amount*o[k2]      #卖出的金额
            w=(price0-1000000)*100    #利润
        Price0.append(0)
        Price.append(w)
        
print("利润：%f"%w)

plt.plot(Price0,color="Black",label="initial")
plt.plot(Price,color="green",label="Profit and Loss")
plt.xlabel("Date")
plt.ylabel("Price")
plt.show()
```
