# 本杰明格雷厄姆成长股投资策略

```python
def stock_sellection_Graham_growth_stock(context):
    """
    操盘控制（预测）
    成长价值=当前（正常）利润*（8.5 + 两倍的预期增长率）
    成长价值 > 大于市值， 买入 反之卖出
    
    """
    # 数据获取（通用部分）
    previous_date = context.previous_date
    # previous_date =  previous_date.strftime('%Y-%m-%d')
    previous_date =  previous_date.strftime('%Y%m%d')
    current_universe = context.get_universe(asset_type='stock', exclude_halt=True) # 获取股票池中的股票
    field_1 = ['secID', 'DilutedEPS', 'EGRO'] # secID 为股票代码 DilutedEPS 稀释每股收益 EGRO：5年收益增长率
    field_2 = ['secID', 'closePrice', 'negMarketValue'] # negMarketValue 为流通市值
    # 数据获取（策略需要数据）
    # DataAPI.MktStockFactorsOneDayGet(tradeDate=previous_date,secID=u"",ticker=u"00000",field=u"",pandas="1")
    GGV_EPS_EGRO = DataAPI.MktStockFactorsOneDayGet(tradeDate=previous_date,
                                                    secID=current_universe,
                                                       ticker=u"",
                                                       field=field_1,
                                                       pandas="1")
    GGV_EPS_EGRO.set_index('secID', inplace=True)
    

    GGV_closePrice = DataAPI.MktEqudGet(secID=current_universe,
                                        ticker=u"",
                                        tradeDate=previous_date,
                                        isOpen="",
                                        field=field_2,pandas="1")

    GGV_closePrice.set_index('secID', inplace=True)
    
    GGV_closePrice['Maketable_Volume'] = GGV_closePrice['negMarketValue']/GGV_closePrice['closePrice'] #计算流通股的数量
    # 股票价值
    GGV_EPS_EGRO['Value'] = GGV_closePrice['Maketable_Volume'] * GGV_EPS_EGRO['DilutedEPS']*(8.5 + 2*GGV_EPS_EGRO['EGRO'])
    under_value = GGV_EPS_EGRO['Value'] - GGV_closePrice['negMarketValue']
    
    buylist = under_value[under_value>0].index
    
    # print(buylist)
    return list(buylist)
    
    
    
```

```python
def trading(buylist, context):
    account = context.get_account('fantasy_account')
    current_universe = current_universe = context.get_universe(asset_type='stock', exclude_halt=True) # 获取股票池中的股票
    security_position = account.get_positions() # 获得持有仓位
    cash = account.cash
    
    for sec in current_universe:
        if sec not in buylist and sec in security_position:
            account.order_to(sec, 0)
    d = min(len(buylist), int(cash)//Max_Position)
    for sec in buylist[:d]:
        if sec not in security_position:
            account.order(sec, Max_Position/context.current_price(sec)) #每只股票最大投资额为1000000
            
```
