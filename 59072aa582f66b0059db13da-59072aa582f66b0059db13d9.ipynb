```python

class Turtle(object):
    ATR = 0.0
    today = ''
    unit = 0
    breakAttr = 0
    stops = []
    unitCount = 0
    lastBreakPrices = 10000
    unit_limit = 5
    hold_flag = False
    breakIngore = False
    daily_data = None
    lastPrices = 0.0
    stk = ''

    def __int__(self):
        init()

    def init(self):
        self.ATR = 0.0
        self.unit = 0
        self.breakAttr = 0
        self.stops = [] 
        self.unitCount = 0
        self.lastBreakPrices = 10000
        self.unit_limit = 4
        self.hold_flag = False
        self.breakIngore = False
        self.lastPrices = 1000.0

    def do_daily(self, account, T, date):
        if self.today != date or  self.daily_data is None:
            self.today = date
            if self.unitCount == self.unit_limit:
                self.unitCount -= 2
            self.daily_data = self.del_tingpai(account, T)
            
            if len(self.daily_data) <= 10 :
                self.inited = False
                return self.daily_data
            else:
                self.inited = True
            
            self.calcATR(date, account)
            self.calcUnit(account)
        return self.daily_data    
        
    def calcUnit(self, account):
        if math.isnan(self.ATR):
            self.inited = False
            return 
        else:
            self.inited = True
        value = account.reference_portfolio_value * 0.01 
        self.unit = int((value/self.ATR)/TCOUNT/100)*100
        return self.unit

    def calcATR(self, date, account):
        high = np.array(self.daily_data['highPrice'])
        low = np.array(self.daily_data['lowPrice'])
        close = np.array(self.daily_data['closePrice'])
        ATR = ta.ATR(high, low, close, timeperiod=20)
        self.ATR = ATR[-1]
        record['ATR'].update({self.today:self.ATR})

        return self.ATR

    # 入市
    def do_admission(self, prices, T):
        if not self.inited:
            return
            
        if self.unitCount >= self.unit_limit:
            return False

        up20 = max(self.daily_data['highPrice'].iloc[-T:])
        up55 = max(self.daily_data['highPrice'].iloc[-55:])
        if prices > up55 and self.breakIngore:
            print self.today,'in up55'
            order(self.stk, self.unit)
            self.unitCount += 1
            self.stops.append(prices-2*self.ATR)
            self.hold_flag = True
            self.lastPrices = prices
            return True
        elif prices > up20:
            if self.lastBreakPrices >= prices: # 上次不是盈利性突破
                print self.today,self.stk, 'in up20'
                self.lastBreakPrices = prices
                order(self.stk, self.unit)
                self.unitCount += 1
                self.stops.append(prices-2*self.ATR)
                self.hold_flag = True
                self.breakIngore = False
                self.lastPrices = prices
                record['break_up'].update({self.today:prices})
                return True
            else:
                print 'ingore'
                self.breakIngore = True

    def do_add(self, prices, close):
        if not self.inited:
            return
            
        if prices > self.lastPrices+0.5*self.ATR and self.unitCount < self.unit_limit:
            order(self.stk, self.unit)
            self.unitCount += 1
            self.hold_flag = True
            self.lastPrices = prices
            self.stops.append(prices-2*self.ATR)
            '''
            for i in range(0, len(self.stops)):
                self.stops[i] = prices - 2*self.ATR
            '''
            print self.today, 'add', self.unitCount, len(self.stops)

    def do_leave(self, prices, T=20):
        if not self.inited:
            return
            
        down = min(self.daily_data['lowPrice'].iloc[-int(T/2):])
        if prices < down:
            print self.today, self.stk, 'leave'
            order_to(self.stk, 0)
            record['break_down'].update({self.today:prices})
            self.init()
            return True
        return False

    def do_stop(self, prices):
        if not self.inited:
            return
            
        if prices < self.stops[-1]:
            print self.today, self.stk, 'stop'
            order_to(self.stk, 0)
            record['stop_loss'].update({self.today:prices})
            self.init()
        '''
        for s in self.stops:
            if prices < s:
                order(stk, -self.unit)
                self.unitCount -= 1
                self.breakAttr = -1
        '''        

    def del_tingpai(self, account, t, act=True):
        act_t = 0
        delta = t - act_t
        act_t += delta
        data = None
    
        while delta > 0:   # 直到取满T个不为停牌的数据
            data = account.get_daily_history(act_t)
            data = data[self.stk]
            data = pd.DataFrame(data)
            if not act:
                break   
            data['highPrice']  = data['highPrice'].replace(0,np.nan)
            data = data.dropna()
            delta = t-len(data)
            act_t += delta
            if act_t > 300:
                return data

        return data
                
```
