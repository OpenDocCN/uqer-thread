## **基于ICA的股票池成分分析**
我们认为股票的价格波动中存在一些潜在的共同因素，了解这些因素可能能揭示一内在的推动机制，从而为股票价格变化提供一些参考。为此本文给出了独立成分分析(ICA)在股票池成分分析中的一种运用方法，为成分分析的后续研究提供一些参考和思路。

## **1，原理分析**
ICA又称盲源分离(Blind source separation, BSS)，它假设观察到的随机信号x服从模型$$\begin{equation}\mathbf{x}=\mathbf{A}\mathbf{s}\end{equation}$$其中s为未知源信号，其分量相互独立，A为一未知混合矩阵。ICA的目的是通过且仅通过观察x来估计混合矩阵A以及源信号s。<br>
<br>
ICA的算法需要用PCA进行降维，同时进行白化和中心化处理。所谓白化就是将数据的方差标准化为1，中心化就是将均值标准化为0,设x为观测向量即$$\begin{equation}E\{xx^T\}=I\end{equation}$$<br>$$\begin{equation}E\{x\}=0\end{equation}$$
<br>
PCA的基本思想是去除数据中共同影响带来的信息冗余。举个例子，在D-2空间中存在一条直线，其参数方程为：
$$\begin{equation}x=t+1\end{equation}$$
$$\begin{equation}y=t+2\end{equation}$$
由于存在共同的影响因素只有一个t，因此这条直线在一维空间上就可以表示出来。
实际情况中我们试图找到少量的几个因素，其对数据的影响最大，从而可以用这几个数据组合得到数据的一个近似，从而达到降维的目的。<br>
<br>
根据上文所述我们要找的影响最大的几个内在因素，那么其数学目标就是找到特征幅度变化最大的方向，这个方向在数学上叫做主方向（“major axis of variation”）。这一点理解很重要。从反面理解，幅度变化最小的方向就是没有变化，或者非常非常小的变化（可以忽略的变化），相对来说可利用价值最小，最可以忽略。而为了找到特征变化最大的方向，假设单位方向矢量为u，则特征点x在u方向的投影点x’距离原点的距离为d=xTu。所有的样本点都在一个方向投影后，他们就都在同一条直线上了。而要比较它们之间变化的程度，只要比较d的方差就行。<br>
<br>
因此我们得到了目标函数：<br>

其中x的上标i表示数据集中的第i个样本，m表示数据集中的样本总数。<br>
<br>
仔细观察上面函数可以发现括号中的部分就是协方差矩阵，于是我们得到了简化后的优化方程：<br>
$$\begin{equation}max\ u^T∑u\end{equation}$$<br>
$$\begin{equation}s.t.\ u^Tu=1\end{equation}$$<br>
<br>
用拉格朗日乘数法求解上面的最大化问题，很容易得到：
$$\begin{equation}∑u=λu\end{equation}$$<br>
这就是协方差矩阵的特征之分解，u为其特征向量，λ为其特征值。因此可以通过协方差矩阵的迹衡量方差的大小。最大的特征值λ（以及对应的特征向量u）决定了数据变化最大的方向。u就是这个单位方向。因此PCA的求解过程就是对协方差矩阵进行特征值分解，并找到最大的几个特征值的过程。然后由最大的k个特征值对应的特征向量组成一组新的基（basis），将原空间映射到新的空间上，降维工作就完成了。<br>
<br>
这是的数据虽然已经是低维的，但其依然可能是源驱动信号的组合，因此我们对低维数据作进一步的ICA去混合处理，这里本文采用的是python的sklearn包。其基本原理是如下步骤：<br>
>1. 初始化随机向量$\ w$；
>2. 设x为观测向量，$w^+ = E{xg(w^Tx)} –  E{g'(w^Tx)}w$；
>3. 令$\ w= w^+/||w^+||$；
>4. 若未收敛，返回2；

其中收敛意味着前后两次向量w在同一方向上，即它们的点积为1。 一次FastICA算法能估计出一个独立成分，为了估计出若干个独立成分，需要进行若干次FastICA算法得到向量$w_1，…，w_n$。

## **4，协方差矩阵建立和分解**

```python
cov=DataFrame(index=universe,columns=universe)
for stk in universe:
    for stk2 in universe:
        cov[stk][stk2]=np.mean(stock_frame[stk]*stock_frame[stk2])-np.mean(stock_frame[stk])*np.mean(stock_frame[stk2])
DataFrame(cov)
```

```python
Cov_mat=np.zeros((50,50))
for i in range(50):
    for j in range(50):
        Cov_mat[i][j]=cov[universe[i]][universe[j]]
Cov_mat
```

## **7，改进和研究方向**
- 由于ICA方法是一种非参数方法，因此有较强的客观性，在金融时间序列中可以有很广泛的用途。其他研究者可以尝试对ICA的结果给出经济解释，并借此对未来价格变动进行预测，本文认为两个信号很有可能分别是相对理性投资者和噪音投资者。
- 另外，拉长时间跨度也是一个选择，对多年的数据进行降维很有可能可以分离出周期性驱动力。<br>
- 但是ICA也存在一些限制，例如它可以很好的解除线性相关，但是对于高阶相关性就没有办法了。对此可以采取的改进方法是对于存在高阶相关性的数据，考虑核ICA，通过核函数将非线性相关转为线性相关。
