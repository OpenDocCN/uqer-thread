# 如何观察数据（假设正态分布）

方法一。[ $mean-1.96\times \sigma,mean+1.96\times\sigma $]。其中mean和std都在上面了

如此问题就变成了一个很好玩的解方程，$$ n\times mean-1.96\sigma \times sqrt(n) > 0 $$

```python
# random walk
mean = 5.991036
std  = 8.988141
for i in np.arange(1,10,0.01):
    if i * mean - 1.96*std*math.sqrt(i)>0: # 这是1.96的含义 stats.norm.ppf(0.975)
        print(i)
        break
```

# 债券基金的回撤周期

```python
def maxdrawdown(df,col_val,col_date):
    """
    maxdrawdown of a security
    :param df: DataFrame of the historical stock data
    :param col_val: column name of df,价格历史数据
    :param col_date: column name of df,日期历史数据,必须为升序
    :return:
    """
    res = df[[col_date,col_val]]
    res['rel_max'] = pd.expanding_max(res[col_val])
    res['drawdown'] = 1 - res[col_val] / res['rel_max']
    maxdd = res['drawdown'].max()

    idx_bottom = res['drawdown'].argmax()
    date_bottom = res.ix[idx_bottom][col_date]  # 最大回撤的低点

    idx_top = res.loc[:idx_bottom, col_val].argmax()
    date_top = res.ix[idx_top][col_date]  # 最大回撤的高点

    return date_top,date_bottom,maxdd 

fundlist = DataAPI.FundGet(etfLof=u"",secID=u"",ticker=u"",category='B',idxID=u"",idxTicker=u"",listStatusCd=u"",operationMode=u"",field=u"",pandas="1")
dd = {}
for secID in fundlist['secID']:
    df = DataAPI.FundNavGet(dataDate=u"",secID=secID,ticker=u"",beginDate=u"",endDate=u"",field=u"",pandas="1")
    df = df[['endDate','ADJUST_NAV']]
    entry = {}
    if not df.empty:
        dt_l,dt_r,maxdd = maxdrawdown(df,'ADJUST_NAV','endDate')
        entry['dt_l']=dt_l
        entry['dt_r']=dt_r
        entry['maxdd']=maxdd
        dd[secID]=entry
```

```python
res = pd.DataFrame(dd).T
```

```python
res['dt_l'] = pd.to_datetime(res['dt_l'])
res['dt_r'] = pd.to_datetime(res['dt_r'])

res['rel_day'] = (res['dt_r'] - res['dt_l']) / np.timedelta64(1, 'D')
res.loc[res['rel_day']!=0,'rel_day'].hist(bins=30,figsize=(12,8))
```
