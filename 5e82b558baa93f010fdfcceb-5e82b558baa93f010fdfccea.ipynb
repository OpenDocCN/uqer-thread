#### 构建季度因子
之前783因为取了inner而消失
先根据intoDate和outDate来选取各张表中的因子，然后再outer merge

```python
# x是整体的值
# start_end_df 表示了开始和结束日期
def select_effective(x, start_end_df):
    temp1 = x.publishDate >= start_end_df.start_date
    temp2 = x.publishDate <= start_end_df.end_date
    temp3 = x.ticker == start_end_df.stockcode
    temp4 = start_end_df.loc[temp1 & temp2 & temp3]
    if len(temp4) == 0:
        return np.nan
    else:
        return temp4.ykcode.values[0]
```

```python
temp_df = pd.read_csv('FdmtIndiGrowthGet2011-2019.csv',encoding='gbk')
start_end_df = pd.read_csv('start_end.csv',encoding='gbk')
temp_df['ykcode']= temp_df.apply(lambda x: select_effective(x,start_end_df), axis = 1)
temp_df = temp_df[pd.isna(temp_df.ykcode)==False]

temp_df1 = pd.read_csv('FdmtIndiRtnGet2011-2019.csv',encoding='gbk')
temp_df1['ykcode']= temp_df1.apply(lambda x: select_effective(x,start_end_df), axis = 1)
temp_df1 = temp_df1[pd.isna(temp_df1.ykcode)==False]

temp_df2 = pd.read_csv('FdmtIndiTrnovrGet2011-2019.csv',encoding='gbk')
temp_df2['ykcode']= temp_df2.apply(lambda x: select_effective(x,start_end_df), axis = 1)
temp_df2 = temp_df2[pd.isna(temp_df2.ykcode)==False]
```

```python
key_list = ['secID','publishDate','endDate','partyID','ticker','ykcode']
new_df = pd.merge(temp_df,temp_df1,left_on = key_list, right_on = key_list,how = 'outer')
new_df = pd.merge(new_df,temp_df2,left_on = key_list,right_on=key_list,how = 'outer')
new_df.reset_index(drop=True,inplace=True)
new_df = pd.merge(new_df,start_end_df,left_on = ['ykcode'],right_on=['ykcode'],how = 'left')
new_df.drop(columns=['ticker'],inplace=True)
new_df.to_csv('Q_temp.csv',encoding='gbk',index=False)
```

```python
temp_df = pd.read_csv('pledgeRatio1.csv', encoding='gbk')
# 在intoDate和outDate之间且ticker相同的，就相加在一起成为质押股东数据
def fun_pledgeRatio(x,df):
    temp1 = x.stockcode == df.ticker
    temp2 = x.publishDate < df.outDate
    temp3 = x.publishDate >= df.intoDate
    temp4 = df.accPledgeTotal[temp1 & temp2 & temp3]
    if len(temp4) == 0:
        return np.nan
    else:
        return temp4.sum()
new_df['pledgeRatio'] = new_df.apply(lambda x: fun_pledgeRatio(x,temp_df), axis = 1)
# temp_df.accPledgeTotal[:3] 前三个！！！
# temp_df.accPledgeTotal.loc[:3] 到index3！！！
```

```python
# 大股东质押数字是%为单位
def fun_ifpledgeRatio_cond(x):
    if np.isnan(x.pledgeRatio):
        return np.nan
    elif np.isnan(x.mny_debtRatio):
        return np.nan
    elif x.pledgeRatio > 90:
        return True
    elif (70<=x.pledgeRatio<=90) & (x.mny_debtRatio < 0.3):
        return True
    else:
        return False
new_df['ifpledgeRatio_cond'] = new_df.apply(lambda x: fun_ifpledgeRatio_cond(x),axis = 1)
new_df.to_csv('Qtemp.csv',encoding='gbk',index=False)
```

```python
def fun_renew(x,list1,str1):
    if np.isnan(x[str1]):
        return np.nan
    elif x.indust_sw in list1:
        return False
    else:
        return x[str1]
    
def fun_renew_com(x,str1):
    if np.isnan(x[str1]):
        return np.nan
    elif x.partyNature1 == '国有企业':
        return False
    else:
        return x[str1]
```

```python
new_df['ifoperatecash_cond'] = new_df.apply(lambda x: fun_renew(x,ifoperatecash_cond_list,str1 = 'ifoperatecash_cond'), axis = 1)
new_df['iftax_cond'] = new_df.apply(lambda x: fun_renew(x,iftax_cond_list1,str1 = 'iftax_cond'), axis = 1)
new_df['iftax_cond'] = new_df.apply(lambda x: fun_renew_com(x,str1='iftax_cond'), axis = 1)
new_df['ifmny_debtRatio_cond'] = new_df.apply(lambda x: fun_renew(x,ifmny_debtRatio_cond_list1,str1 = 'ifmny_debtRatio_cond'), axis = 1)
new_df['ifmny_debtRatio_cond'] = new_df.apply(lambda x: fun_renew_com(x,str1='ifmny_debtRatio_cond'), axis = 1)
```
