# 做多波动率
当然，做事件的策略，如果不追求特别多的收益，那做多波动率是个不错的选择。

我们来一起做一下复盘，看看几种做多波动率的方式的收益如何。

如果可以操作外汇期权，那是极好的。如果不能，那我们一起来看看国内的50ETF期权的情况。

选择下面三种操作方式：

* 在星期四买入2.10行权价的近月期权，用认沽和认购合约组成跨式多头，在周五收盘卖出。
* 在星期一买入2.10行权价的近月期权，用认沽和认购合约组成跨式多头，在周五收盘卖出。
* 在星期一买入2.10行权价的12月期权，用认沽和认购合约组成跨式多头，在周五收盘卖出。

## Plan A 在星期四买入2.10行权价的近月期权，在周五收盘卖出。

```python
from CAL.PyCAL import *
from pandas import concat
import numpy as np
from scipy import interpolate
import pandas as pd

import lib.options_watch_dog as WatchDog

opt_thursday = WatchDog.TTable_Options_50ETF(Date(2016,6,23), column_type=2)
opt_near = opt_thursday[opt_thursday.index=='2016-07-27']
opt_near
```

```python
price_call_07 = opt_near.iloc[3][u'认购合约'][u'收盘价']
price_put_07 = opt_near.iloc[3][u'认沽合约'][u'收盘价']
total_paid = price_call_07 + price_put_07
print u'6月23日，收盘时。认购价格：%s，认沽价格：%s，总支出：%s' % (price_call_07, price_put_07, total_paid)
```

```python
opt_friday = WatchDog.TTable_Options_50ETF(Date(2016,6,24), column_type=2)
opt_near = opt_friday[opt_friday.index=='2016-07-27']
opt_near
```

```python
price_call_07 = opt_near.iloc[3][u'认购合约'][u'收盘价']
price_put_07 = opt_near.iloc[3][u'认沽合约'][u'收盘价']
total_paid = price_call_07 + price_put_07
print u'6月24日，收盘时。认购价格：%s，认沽价格：%s，总收入：%s' % (price_call_07, price_put_07, total_paid)
```

根据上面两个节点的计算，这次交易的收益为 (0.0884-0.0883)/0.0883=0.0011

也就是说，千分之一。收益很有限。

# 我想要更多
当然，想要更多收益，那就不能在收盘价时候平仓了。而是要在关键时刻操作。

这个关键时刻当然就是周五下午的一点到两点。此时的波动率最大。做多波动率，当然要在波动率最大的时候平仓。

看下面图片

![激动时刻](http://o9cyuhwvg.bkt.clouddn.com/2016-06-25_2107.png)

如果你当时有在看盘，会看到，中间有一次波动率中断，也就是交易熔断，当交易恢复后，价格达到最低点。

此时挂单成交，可以获得更多收益。

此时，波动率最高点。认购价格：0.0316，认沽价格：0.072，总收入：0.1036

收益为 (0.1036-0.0842)/0.0842 = 0.230

# 总结
* 做多波动率，早点下手。
* 日内波动率往往会增加，但收盘时可能又会回落。
* 平仓时，要注意择时
