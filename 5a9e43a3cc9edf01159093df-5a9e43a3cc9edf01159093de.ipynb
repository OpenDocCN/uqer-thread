# 50etf & hsetf corr test

```python
#从api获取历史数据
import pandas as pd
field = [u"tradeDate", u"closePrice"]
sz50 = DataAPI.MktFunddGet(tradeDate=u"",secID=u"510050.XSHG",ticker=u"",beginDate=u"20170101",endDate=u"20180101",field=field,pandas="1").set_index('tradeDate')
sz50.index = pd.to_datetime(sz50.index)
hsetf = DataAPI.MktFunddGet(tradeDate=u"",secID=u"159920.XSHE",ticker=u"",beginDate=u"20170101",endDate=u"20180101",field=field,pandas="1").set_index('tradeDate')
hsetf.index = pd.to_datetime(hsetf.index)
sz50.columns = ["sz50Close"]
hsetf.columns = ["hsetfClose"]
print "==================="
```

```python
#合并两个指数到同一个DataFrame
data = sz50.join(hsetf)
```

```python
data = data.dropna()
data.tail()
```

```python
#计算Pearson相关系数
data.corr()
```

相关性	负	正
无	−0.09 to 0.0	0.0 to 0.09
弱	−0.3 to −0.1	0.1 to 0.3
中	−0.5 to −0.3	0.3 to 0.5
强	−1.0 to −0.5	0.5 to 1.0

```python
#看看更多分析，寻找交易机会
data.describe()
```

```python
#每20天计算的相关系数
corr = pd.rolling_corr(sz50["sz50Close"], hsetf["hsetfClose"], 30, min_periods=20)
corr.plot(legend=True)
```

```python
var1 = pd.rolling_var(hsetf["hsetfClose"], 30 , min_periods=20)
var1.plot()
```

```python
var2 = pd.rolling_var(sz50["sz50Close"], 30 , min_periods=20)
var2.plot()
```

```python
#计算收益率
ret = data.apply(lambda x: x/x.shift(1)-1)
ret = ret.dropna()
ret.tail()
```
