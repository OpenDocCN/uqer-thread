# 基金回测框架中工作日及其前后日期的处理

之前做基金回测框架的时候涉及到一个工作日的处理问题，虽然是个小问题，不过也挺烦人的，写了个小程序，分享一下:)

```python
# 工作日列表
workday = [date.replace('-','') for date in DataAPI.FundNavGet(ticker='000001',beginDate='20050101',endDate='20990101',field=u"endDate",pandas="1").endDate]

def current_date(date):
# 计算当前所属工作日
    if date in workday:
        current_date = date
    else:
        for d in workday:
            if d > date:
                current_date = d
                break
    return current_date

def next_day(date):
    return workday[workday.index(current_date(date))+1]

def last_day(date):
    return workday[workday.index(current_date(date))-1]

def date_between(start_date, end_date):
    # 不包括开始和结束日期
    return workday[workday.index(current_date(start_date))+1: workday.index(current_date(end_date))]
        
# 测试    
if __name__ =='__main__':
    print current_date('20150101')
    print next_day('20150101')
    print last_day('20150101')
    print date_between('20141201','20150120')
 
```
