# 『量化投资这一年』一年的优矿生涯的总结

**0.开篇**

$\quad\quad$我是一名程序员，从大一到研三都是偏向计算机方向的，到了工作也是从事计算机IT方向的工作。接触量化和优矿几乎是同时的，那时还在CFA考试中，很想将自己所学的知识实战一下，很自然的想到了量化，于是也第一时间找到了优矿。一接触可以说就有点“沉迷”，其实我还是很喜欢研究生的生活，但是毕业以后因为工作的原因，真正攻关开发的时间往往很少，加班的时间也不多，所以还是有很多闲置的时间吧。量化研究很像搞科研，你要查阅文档（国内的买方卖方的研报，国外杂志上的论文，国内的论文等等）；通读相关方向上的文档之后，还要试着理解文章中所讲得东西，并尝试提出改进或者新的想法；之后对于研报的东西，写程序进行实现，看看策略实际效果如何，对于新的想法，写程序则看看有没有实际的突破；最后便是文字整理，将技术细节整理成可读的文档，并配上相应的图表，如果策略创新失败那就总结失败经验，提出进一步改进措施和接下去的工作提案。总的来说，进入资本市场是因为自己的好胜心，好奇心与智力自信，学习量化是因为我的编程算法技能与量化技术的贴近。话说，我还真是fall in love with it！

**1.优矿**

$\quad\quad$下面再谈谈优矿。
$\quad\quad$优矿有一个很好很活跃的社区，这令我接触量化知识非常方便。带我入门的便是优矿的《量化分析师Python日记》系列，在大学里我用的最多的是Matlab，对于Python不是很熟悉，所以也是从头学起，好在有着C语言的基础，Python学习也不是很困难，经过大约一周的学习基本可以写一些简单的策略，但是遇到生僻的数据结构的用法还是要求优矿与百度帮忙。
$\quad\quad$在语言关合格之后，优矿社区中的精华策略贴帮助我了解到了一些常用的策略与量化投资知识。如果只是单单看券商研报的话，很多细节的部分就会不求甚解，但是有了优矿上的帖子之后，便会对一些实现的细节豁然开朗，然后与之前所用到的知识相互融汇贯通,得出一些或正确或错误的想法，然后便可以在优矿实验平台上进行验证；每次有新的想法的时候都很兴奋，希望能在原来的基础上更进一步，虽然常常在回测阶段就被证明是无效的想法，但是成功的喜悦就像买到一支10倍股一样，会将之前的被资本市场教训所带来的沮丧消弭殆尽。同时优矿还提供了模拟平台，可以更加真实的检测自己的策略，看看策略是否能收益稳定，估计受到的冲击成本的大小和资金容量。

**2.策略**

$\quad\quad$然后再谈谈量化策略，通过一年的学习，我尝试了不少大大小小不下30余种策略，也算是有一点点小小的心得。
$\quad\quad$相比基本面策略来说，量化分析的优势在于分析个股较快，能够快速的扫遍资本市场的各个股票，分析3000多只股票只是一个点击的功夫，但是对于基本面分析来说一支股票至少要几个小时的分析，所以量化选股能够在最小的人力情况下分析更多的股票，这个特性决定了量化选股可以通过提高换手率获得更高的收益（多下注，通过大数定律取胜），所以也更加适合高频投资。
$\quad\quad$而量化的劣势在于对于个股的研究深度有限，只能针对一些可以量化的因素，相较于基本面来说像行业分析这样的需要一点“灵气”的分析，量化考虑问题的深度往往不够，选股的质量也稍差，这一点是通过多下注，选多股来平衡。同时由于量化分析通常基于统计归纳的结果，而不是逻辑演绎，多少带上一些无法说清基本原理的黑盒因素（和非理性行为一样）；不过随着数据库的完善，因子库的增加，非线性选股工具（机器学习，深度学习）的出现，量化分析覆盖的深度会逐渐增加，这时分析广度和人力成本的优势也会逐渐突出。
$\quad\quad$相对于技术分析来说，量化分析可以代替基本上所有的量价指标模型，随着模式识别技术的发展，在形态上的分析技术也会逐渐先进，像波峰波谷，十字星的等形态判断也会越来越准确，往往能取到取代技术分析的效果。
$\quad\quad$我想随着中国市场的完善，数据库的丰富，T+0与衍生品的逐渐推出，中国的量化投资市场也会逐渐水涨船高！

**3.结束**

$\quad\quad$18年，我想会做一些期货和中性策略并尝试一下不同品种标的；18年，我在优矿上的量化研究还会继续，无惧回撤，收获超额收益！
