# 配对交易（示例6期货套利）

最近，期货市场非常红火，黑色产业链各品种更是充满了投资机会，虚拟钢厂套利就是一个很好的低风险交易策略。

```python
from matplotlib import pylab
import numpy as np
import pandas as pd
import DataAPI
import seaborn as sns
sns.set_style('white')

#设置期货合约代码
secID1 = 'I1609.XDCE'
secID2 = 'RB1610.XSHG'
secID3 = 'J1609.XDCE'
#设置计算周期
start = '20151101'
end = '20160412'
#获取期货合约行情数据
security1 = DataAPI.MktFutdGet(secID = secID1, beginDate=start, endDate=end, field=['tradeDate', 'closePrice'])
security2 = DataAPI.MktFutdGet(secID = secID2, beginDate=start, endDate=end, field=['tradeDate', 'closePrice'])
security3 = DataAPI.MktFutdGet(secID = secID3, beginDate=start, endDate=end, field=['tradeDate', 'closePrice'])
s_temp = pd.merge(security1,security2,on = 'tradeDate')
security = pd.merge(s_temp,security3,on = 'tradeDate')

#按照公式螺纹钢/（铁矿石*1.6+焦炭*0.45）设置虚拟钢厂
security['vProfit'] = security['closePrice_y']/(security['closePrice_x']*1.6+security['closePrice']*0.45)
security['v_m']=security['vProfit'].mean()


security['v_m_up'] = security['v_m']+np.std(security['vProfit'])
security['v_m_down'] = security['v_m']-np.std(security['vProfit'])

security['tradeDate'] = pd.to_datetime(security['tradeDate'])
security = security.set_index('tradeDate')
security.info()

security['vProfit'].plot(grid=False, figsize=(12,8))
security['v_m'].plot(grid=False, figsize=(12,8))
security['v_m_up'].plot(grid=False, figsize=(12,8))
security['v_m_down'].plot(grid=False, figsize=(12,8))

sns.despine()

```
