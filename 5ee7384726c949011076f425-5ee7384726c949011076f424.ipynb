# 历史数据的获取

# history获取历史数据
## context.history(symbol, attribute=['closPrice'], time_range=1, freq='1d', style='sat', rtype='frame')
获取指定证券的历史行情、因子等时间序列数据。
symbol：     需要获取数据的证券列表，支持单个证券或证券列表，需要是universe中存在的证券。返回  str或者list。
attribute：  需要获取的属性，支持单个属性或属性列表。                               返回  str或者list。
time_range： 需要回溯的历史K线条数，和freq属性相对应。日线数据默认最大值为30，分钟线数据默认最大值为240；可以使用                    max_history_window设置最大限度取值范围。                                         参数类型int。
freq：       K线图周期。参数类型str。参数用法：日线K线图：1d。分钟线K线图：1m，5m，15m，30m，60m。
style：      数据返回的格式，可以选择ast 、sat、 tas 三种。a表示attribute，s表示symbol，t表示time，
             三种选择分别对应着这三个维度呈现的顺序，
             例如ast就表示返回的字典中的键是attribute，其值是列为symbol、行为time的DataFrame，其他以此类推。
             参数类型str。
             可以同时取多个品种资产对应的所有支持数据，包含期货品种时数据返回类型必须为style = 'sat'
rtype：      返回值的数据类型。参数类型：str，支持两个值：'frame'，'array'。
             返回dict，key为资产符号，value的格式由rtype决定。
