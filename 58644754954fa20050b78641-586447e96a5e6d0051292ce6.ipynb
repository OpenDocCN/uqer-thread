```python
from CAL.PyCAL import *
from pandas import DataFrame
import numpy as np
import pandas as pd



start = '2012-08-01'
end = '2015-08-01'
benchmark = 'HS300'
universe = set_universe("HS300")
capital_base = 1000000
refresh_rate = 20


#######################    唯一需要自定义的核心代码：构建信号    #######################  
def foo1(data, dependencies=['turnoverVol'], max_window=5):
    signal_5vol = data['turnoverVol'].mean()
    return signal_5vol
def foo2(data, dependencies=['closePrice'], max_window=60):
    yesterday_price = data['closePrice'].ix[-1]  # 获取上个交易日
    day60_price = data['closePrice'].ix[-60]   
    signal_revs60 = (yesterday_price/day60_price)-1
    return signal_revs60

#######################    唯一需要自定义的核心代码：构建信号    #######################      
#回测区间设置为大赛官方统一的标准，起始时间从2012年8月1日至2015年8月1日，起始资金为100万

def initialize(account):
    a = Signal("a", foo1)
    b = Signal("b", foo2)
    account.signal_generator = SignalGenerator(a,b)
    
def handle_data(account):
    wo_yao_can_sai(account)
    
# 我要参赛函数
def wo_yao_can_sai(account):
    yesterday = account.previous_date.strftime('%Y%m%d')
    c = a+b
    # 信号获取
    sig = account.signal_result['c']
    sig = sig.replace([np.inf, -np.inf], np.nan).dropna()
    sig = dict(sig)

    # 合规处理
    univ = sig.keys()

    # 1. 去除停牌股票
    univ = [s for s in univ if s in account.universe]

    # 2. 去除ST股
    try:
        STlist = DataAPI.SecSTGet(secID=univ, beginDate=yesterday, endDate=yesterday, field=['secID']).tolist()
        univ = [s for s in univ if s not in STlist]
    except:
        pass

    # 3. 去除流动性差的股票
    tv = account.get_attribute_history('turnoverValue', 20)
    mtv = {sec: sum(tvs)/20. for sec, tvs in tv.iteritems()}
    univ = [s for s in univ if mtv.get(s, 0) >= 10**7]

    # 4. 去除新上市或复牌的股票
    opn = account.get_attribute_history('openPrice', 1)
    univ = [s for s in univ if not (np.isnan(opn.get(s, 0)[0]) or opn.get(s, 0)[0] == 0)]

    # 因子处理
    sig = {s: sig[s] for s in univ}
    
    # 获得权重
    wts = simple_long_only(sig, yesterday)

    # 卖掉不在权重字典当中的股票
    for stock in account.valid_secpos:
        if stock in account.universe and stock not in wts:
            order_to(stock, 0)

    # 根据权重计算调仓数量
    c = account.referencePortfolioValue
    change = {}
    for stock, w in wts.iteritems():
        p = account.referencePrice[stock]
        if p and not np.isnan(p):
            change[stock] = int(c * w / p) - account.valid_secpos.get(stock, 0)

    # 按先卖后买的顺序发出指令
    for stock in sorted(change, key=change.get):
        order(stock, change[stock])
```
