### 作业一

```python
def f(x, sign=1):
    return sign * (2 * x[0] * x[1] + 2 * x[0] - x[0] ** 2 - 2 * x[1] ** 2)
```

```python
x = np.arange(-5, 5, 0.1)
y = np.arange(-5, 5, 0.1)
x, y = np.meshgrid(x, y)
z = f([x, y])
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
```

```python
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.view_init(elev=0., azim=90) # 改变绘制图像的视角,即相机的位置,azim表示转动z轴，elev表示转动x轴
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
surf = ax.plot_surface(x, y, z, cmap=cm.coolwarm, linewidth=0., antialiased=True)
fig.colorbar(surf, shrink=0.5, aspect=5)
```

```python
def f_jac(x, sign=1.0):
    dfdx = sign*(2*x[1] + 2 - 2*x[0])
    dfdy = sign*(2*x[0] - 4*x[1])
    return np.array([dfdx, dfdy])
```

```python
x0 = np.array([0, 0])
result_wo_cons = optimize.minimize(f, x0, args=(-1.0,), jac=f_jac, method='SLSQP').x
print "x: %s, y: %s" % (result_wo_cons[0], result_wo_cons[1])
```

```python
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.view_init(elev=90, azim=0) # 改变绘制图像的视角,即相机的位置,azim表示转动z轴，elev表示转动x轴
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
surf = ax.plot_surface(x, y, z, cmap=cm.coolwarm, linewidth=0., antialiased=True)
ax.scatter(result_wo_cons[0], result_wo_cons[1], f(result_wo_cons), c='y')
fig.colorbar(surf, shrink=0.5, aspect=5)
```

增加约束条件

1. $2x-y=0$
2. $y-1\geqslant0$

```python
cons = (
    {'type': 'eq', 'fun': lambda x: 2 * x[0] - x[1]},
    {'type': 'ineq', 'fun': lambda x: x[1] - 1}
)
```

```python
x0 = np.array([0, 0])
result_wi_cons = optimize.minimize(f, x0, args=(-1.0,), method='SLSQP', constraints=cons).x
print result_wi_cons
```

```python
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.view_init(elev=90, azim=0) # 改变绘制图像的视角,即相机的位置,azim表示转动z轴，elev表示转动x轴
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
surf = ax.plot_surface(x, y, z, cmap=cm.coolwarm, linewidth=0., antialiased=True)
ax.scatter(result_wi_cons[0], result_wi_cons[1], f(result_wo_cons), c='y')
fig.colorbar(surf, shrink=0.5, aspect=5)
# 绘制约束条件
ax.plot(np.arange(-2, 2, 0.1), 2 * np.arange(-2, 2, 0.1)) # 条件1
x_1 = np.arange(-5, 5, 0.1)
y_1 = 1
z_1 = np.arange(-140, 20, 1)
x_1, z_1 = np.meshgrid(x_1, z_1)
ax.plot_surface(x_1, 1, z_1, ) # 条件2
```

$r_p - r_f = \beta \times (r_b - r_f) + \alpha$

```python
rf = 0.03 / 250 # 无风险收益率
def target_func(x): # x0: beta, x1: alpha
    x_beta = x[0] * (benchmark_return_series - rf)
    eps = equity_return_series - rf - x_beta - x[1]
    delta = sum((eps) ** 2)
    return delta

optimize.minimize(target_func, np.array([1, 1]), method='SLSQP', options={'xtol': 1e-1}).x
```

```python
from scipy import stats
beta, alpha, _, _, _ = stats.linregress(benchmark_return_series - rf, equity_return_series - rf)
print beta, alpha
```

```python
A = np.vstack([benchmark_return_series - rf, np.ones(len(benchmark_return_series))]).T
beta, alpha = np.linalg.lstsq(A, equity_return_series - rf)[0]
print beta, alpha
```
