```python

import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from CAL.PyCAL import *
from sklearn import preprocessing
from sklearn import cross_validation
from sklearn import datasets
from sklearn import svm
import itertools
from sklearn.linear_model import ElasticNet
from sklearn.linear_model import Lasso
from sklearn.metrics import f1_score
class MachineToy:
    """
    这部分和上一篇中假设场景有些许差异，上一篇假设在收盘当日运行，而这一篇假设在第二天早晨运行程序
    """
    forward = 0;lag = 0;stock = 0;downUp=0;percent = 0; maxLag = 0; close = True###maxLag为隐藏变量
    features = []
    label = []
    indexDate = pd.DataFrame()
    featureOriginal = pd.DataFrame()
    label_df = pd.DataFrame()
    def __init__(self, forward=3, lag=3, stock = '000001', downUp = 1, percent = 0, close1 = True, autoPrepare = 1):
        """
        forward   :前看天数，默认前看3天
        lag       :特征滞后天数，默认以前3天数据作为特征
        stock     :预测的指数,默认为上证综指
        downUp    :预测跌呢还是跌呢还是跌呢？跌填 0,涨填  1，默认测涨
        percent   :预测幅度，0.02意思就是涨（跌）2个点才标记为正样本
        close1    :计算第二天的涨跌幅是否以明天的开盘价作为起点，不是则以今天收盘价为起点 
        autoPrepare :是否在类实例化的时候就自动准备好features label
        """
        self.forward = forward
        self.lag = lag
        self.stock = stock
        self.percent = percent
        self.close1 = close1
        self.downUp = downUp
        try:
            fields = ['tradeDate', 'closeIndex', 'openIndex', 'highestIndex','lowestIndex', 'turnoverVol']
            indexData = DataAPI.MktIdxdGet(ticker=stock,beginDate=u"19910101",endDate=u"20161212",field=fields,pandas="1")
            if indexData.empty:
                print "获取数据为空"
                return
        except:
            print '无法获取指数数据'
            return
        self.indexData = indexData
        if(autoPrepare ==1):
            self.getFeature()
            self.getLabel()
    def getFeature(self, names = 0):
        featureOriginal = pd.DataFrame()
        data = self.indexData
        featureOriginal['high_low'] = (data['highestIndex'] - data['lowestIndex'])/data['lowestIndex']
        featureOriginal['high_close'] = (data['highestIndex'] - data['closeIndex'])/data['closeIndex']
        featureOriginal['close_low'] = (data['closeIndex'] - data['lowestIndex'])/data['lowestIndex']
        featureOriginal['close_open'] = (data['closeIndex'] - data['openIndex'])/data['openIndex']
        volumeChange = np.diff(np.log(data['turnoverVol']))
        featureOriginal = pd.concat([featureOriginal, pd.DataFrame(volumeChange, columns = ['volumeChange'])], axis = 1)
        featureOriginal['volumeChange'] = featureOriginal['volumeChange'].shift(1)
        featureOriginal = featureOriginal.fillna(0)
        ######形成特征矩阵
        features = []
        for i in range(self.lag, len(featureOriginal) +1):
            temp = featureOriginal[i-self.lag:i]
            temp = list(itertools.chain.from_iterable(temp.values.tolist()))
            features.append(temp)
        if(self.close1 == True):
            features = features[:len(features)-self.forward]
        else:
            features = features[:len(features)-self.forward]
        self.features = features
        featureOriginal['date'] = data['tradeDate']
        self.featureOriginal = featureOriginal
        return
    def getLabel(self):
        """
          
        """
        index= self.indexData
        if(self.close1==False):#以今天的开盘价为起点, 则需将日期推迟2天
            logChange = np.log(index['closeIndex'].shift(-self.forward+1)) - np.log(index['openIndex'])
            index = pd.concat([index, logChange], axis = 1)
            index.rename(inplace =True, columns = {0:'logChange'})
            label_df = pd.DataFrame(index[['tradeDate', 'logChange']])
            label_df['symbol'] = 0
            if(self.downUp==1):
                label_df['symbol'] = label_df['logChange'].apply(lambda x:1 if x>self.percent else 0)         ######symbol 就是我们要的label
            if(self.downUp==-1):
                label_df['symbol'] = label_df['logChange'].apply(lambda x:1 if x<self.percent else 0)
            label_df = label_df[self.lag:len(label_df)-self.forward +1]
        else:  #以昨天的收盘价为起点, 则需将日期推迟3天
            logChange = np.log(index['closeIndex'].shift(-self.forward)) - np.log(index['closeIndex'])
            logChange.name = 'logChange'
            index = pd.concat([index,logChange ], axis = 1)
            label_df = index[['tradeDate', 'logChange']]
            label_df['symbol'] = 0
            if(self.downUp==1):
                label_df['symbol'] = label_df['logChange'].apply(lambda x:1 if x>self.percent else 0)         ######symbol 就是我们要的label
            if(self.downUp==-1):
                label_df['symbol'] = label_df['logChange'].apply(lambda x:1 if x<self.percent else 0)
            label_df = label_df[self.lag - 1:len(label_df)-self.forward]
        self.label_df = label_df
        self.label =  label_df['symbol'].values.tolist()
        
        return
    def train(self, kernel = 'rbf',method = 'lasso'):
        """
        kernal    :核函数，可选择  ‘linear’, ‘poly’, ‘rbf’, ‘sigmoid’, ‘precomputed’，默认为rbf
        """
        features = self.features
        label = self.label
        X_train,X_test, y_train, y_test = cross_validation.train_test_split(features, label, test_size = 0.3, random_state = 0)
        
        if(len(label) == len(features)):
            scaler = preprocessing.StandardScaler().fit(X_train)
            X_train_scaler = scaler.transform(X_train)
            X_test_scaler = scaler.transform(X_test)
            clf = svm.SVC(kernel = kernel).fit(X_train_scaler, y_train)
            myscore = clf.score(X_test_scaler, y_test)
            y_pre = clf.predict(X_test_scaler)
            f1 = f1_score(y_test, y_pre)
            print "准确率：%f, F1: %f"%( myscore,f1)
            return myscore
    
        else:
            print "特征矩阵，标签行数不匹配"
            return
    def plot(self):
        pass
if __name__=="__main__":
    aa = MachineToy(downUp=1,lag=2, forward=3, close1 = True,percent = 0.008)
    aa.train()
    aa.getFeature()
    aa.getLabel()
    kernels = ['linear','poly', 'rbf','sigmoid']
    percents = np.linspace(0,0.03, 7)
    for percent in percents:
        aa = MachineToy(downUp=1,lag=3, forward=2, close1 = True,percent = percent)
        score = aa.train(kernel)


    
    #features = aa.features
    #print features
        



```
