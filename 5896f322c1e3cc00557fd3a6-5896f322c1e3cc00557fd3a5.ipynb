# 秩相关序数和卡方检查的区别

在拜读社区大神 @taotao.li_研究 | 因子分析模版, 谢 call 神 ^_^ 作品时，对秩相关序数的计算有疑虑。
百度百科解释：秩相关系数又称等级相关系数，或顺序相关系数，是将两要素的样本值按数据的大小顺序排列位次，以各要素样本值的位次代替实际数据而求得的一种统计量。
粗略来看，和卡方检验类似，也是用来检查线性关系的，非线性感觉上来讲不大适用
不妨在此做个小测试，测试下秩相关序数和卡方检查在数据上的输出特点
1，构造y=x函数，用秩相关序数检查，用卡方检查
2，y=x+1，分别秩相关序数，开发检查
3，y=x^2正数部分，负数部分，整体，用两者检查

```python
import time
import json
import pandas as pd
import numpy as np
import datetime as dt
import scipy.stats as st
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('white')
from CAL.PyCAL import *
import numpy as np
import itertools
import copy


pd_xy=pd.DataFrame(columns=['x','y'])

#y=x
pd_xy['x']=np.arange(1,1000)
pd_xy['y']=pd_xy['x']
print pd_xy.head()

ic, p_value = st.spearmanr(pd_xy['x'], pd_xy['y'])
chisq, p = st.chisquare(pd_xy['x'], pd_xy['y'])  
print 'y=x'
print 'ic:',ic,p_value,
print 'chis',chisq, p
```

```python
#y=x+1
pd_xy=pd.DataFrame(columns=['x','y'])
pd_xy['x']=np.arange(1,1000)
pd_xy['y']=pd_xy['x']+1
print pd_xy.head()

ic, p_value = st.spearmanr(pd_xy['x'], pd_xy['y'])
chisq, p = st.chisquare(pd_xy['x'], pd_xy['y'])  
print 'y=x+1'
print 'ic:',ic,p_value,
print 'chis',chisq, p
```

```python
#y=x^x
pd_xy=pd.DataFrame(columns=['x','y'])
pd_xy['x']=np.arange(-100,100)
pd_xy['y']=pd_xy['x']*pd_xy['x']
print pd_xy.head()

print 'y=x^2'
ic, p_value = st.spearmanr(pd_xy['x'], pd_xy['y'])
chisq, p = st.chisquare(pd_xy['x'], pd_xy['y'])  
print 'ic:',ic,p_value,
print 'chis',chisq, p

print 'y=x^2 and x<0'
ic, p_value = st.spearmanr(pd_xy['x'][0:100], pd_xy['y'][0:100])
chisq, p = st.chisquare(pd_xy['x'][0:100], pd_xy['y'][0:100])  
print 'ic:',ic,p_value,
print 'chis',chisq, p

print 'y=x^2 and x>0'
ic, p_value = st.spearmanr(pd_xy['x'][100:], pd_xy['y'][100:])
chisq, p = st.chisquare(pd_xy['x'][100:], pd_xy['y'][100:])  
print 'ic:',ic,p_value,
print 'chis',chisq, p
```

y=x

y=x+1
ic: 1.0 0.0 chis 6.48547086055 1.0

y=x^2
ic: -0.0149259272835 0.833844779584 chis nan nan
y=x^2 and x<0
ic: -1.0 0.0 chis 348550.0 0.0
y=x^2 and x>0
ic: 1.0 0.0 chis nan nan

简单来看
y=x：双方都完美表现
y=x+1：ic完美。卡方，第二个p=1.0不错，第一个6.4854的不明含义。也算不错吧
y=x^2:非线性关系，先负相关在正相关，ic表现尚可，第一项基本是0。卡方，两个nan，不晓得为何没返回值。
y=x^2 and x<0：ic检查出负相关关系。卡方，第二项p=0.0，负相关没检查出来，第一项348550这么大，什么鬼。。
y=x^2 and x>0：ic表现ok。卡方，又是两个nan，

感觉上，ic较卡方检查的适用面更广
