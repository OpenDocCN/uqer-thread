using the turnover rate to describe Over-Confidence and cgo score to describe the Disposition Effect

```python
## the functions
def get_tickers_name_by_monthdate(b_date, e_date):
    index_tickers = DataAPI.IdxCloseWeightGet(secID=u"",ticker=u"000905",beginDate= b_date , endDate= b_date, field=['consTickerSymbol', 'weight' ],pandas="1")
    index_tickers = index_tickers.set_index('consTickerSymbol')
    # weight = DataAPI.MktEqudGet(secID=u"",ticker= index_tickers.index.tolist() ,tradeDate= b_date, beginDate=u"",
    #                             endDate=u"",isOpen="",field=['ticker', 'marketValue'],pandas="1")
    # weight.set_index('ticker', inplace = True)
    return index_tickers.index.tolist(), index_tickers['weight'] #weight['marketValue']

def get_basic_line(date):
    index_tickers = DataAPI.IdxCloseWeightGet(secID=u"",ticker=u"000905",beginDate= date , endDate= date, field=['consTickerSymbol', 'weight' ],pandas="1")
    index_tickers = index_tickers.set_index('consTickerSymbol')
    tickers = index_tickers.index.tolist()
    info = DataAPI.MktEqumGet(secID=u"",ticker=tickers ,monthEndDate=u"",beginDate= date,
                                endDate= date, isOpen=u"",field=['ticker', 'chgPct'],pandas="1")
    info.set_index('ticker', inplace = True)
    return   (info['chgPct'] * (index_tickers['weight'] / index_tickers['weight'].sum())).sum()

def get_tickers_turn_by_name_date(tickers, b_date, e_date):
    b_info = DataAPI.MktEqumGet(secID=u"",ticker=tickers ,monthEndDate=u"",beginDate= b_date,
                                endDate= b_date, isOpen=u"",field=['ticker', 'turnoverRate'],pandas="1")
    e_info = DataAPI.MktEqumGet(secID=u"",ticker=tickers ,monthEndDate=u"",beginDate= e_date,
                                endDate= e_date, isOpen=u"",field=['ticker', 'chgPct'],pandas="1")
    info = pd.merge(b_info, e_info, on = 'ticker', how= 'inner')
    return info.set_index('ticker')

def get_tickers_mom_by_name_date(tickers, b_date, e_date):
    p_date = pre_cal[cal['calendarDate'] == b_date]['calendarDate'].values[0]
    day_df = DataAPI.MktEqudGet(secID=u"",ticker=tickers ,tradeDate=u"",beginDate=p_date,
                                endDate=b_date ,isOpen="",field=['ticker', 'closePrice', 'accumAdjFactor', 'turnoverRate'],pandas="1")
    day_df['closePrice'] = day_df['closePrice'] * day_df['accumAdjFactor']
    def compute_cgo(a):
        n = a.shape[0] - 1
        ct = np.array([a['turnoverRate'].iloc[n - i]*  (1 - a['turnoverRate'].iloc[n - i].prod()) * a['closePrice'].iloc[i] for i in range(a.shape[0])])
        cgo = 1 - ct.sum() / a['closePrice'].iloc[-1]
        return cgo
    
    cgo_dict = {i : compute_cgo(day_df[day_df['ticker'] == i]) for i in day_df['ticker'].unique()}
    e_info = DataAPI.MktEqumGet(secID=u"",ticker=tickers ,monthEndDate=u"",beginDate= e_date,
                                endDate= e_date, isOpen=u"1",field=['ticker', 'chgPct'],pandas="1")
    e_info = e_info.set_index('ticker')
    e_info['cgo'] = pd.Series(cgo_dict)
    return e_info


def get_turn_return_info(date):
    e_date = date
    b_date = pre_cal[cal['calendarDate'] == e_date]['calendarDate'].values[0]
    tickers, w = get_tickers_name_by_monthdate(b_date , e_date)
    df_info = get_tickers_turn_by_name_date(tickers, b_date , e_date)
    low_turn = df_info.sort_values('turnoverRate')[:100]
    high_turn = df_info.sort_values('turnoverRate')[-100:]
    low_r = (low_turn['chgPct'] * (w.loc[low_turn.index] / w.loc[low_turn.index].sum())).sum()
    high_r = (high_turn['chgPct'] * (w.loc[high_turn.index] / w.loc[high_turn.index].sum())).sum()
    # low_r = low_turn['chgPct'].mean()
    # high_r = high_turn['chgPct'].mean()
    return (b_date, low_r, high_r), (high_turn.index, low_turn.index)


def get_mom_return_info(date):
    e_date = date
    b_date = pre_cal[cal['calendarDate'] == e_date]['calendarDate'].values[0]
    tickers, w = get_tickers_name_by_monthdate(b_date , e_date)
    df_info = get_tickers_mom_by_name_date(tickers, b_date , e_date)
    low_turn = df_info.sort_values('cgo')[:100]
    high_turn = df_info.sort_values('cgo')[-100:]
    print(date)
    low_r = (low_turn['chgPct'] * (w.loc[low_turn.index] / w.loc[low_turn.index].sum())).sum()
    high_r = (high_turn['chgPct'] * (w.loc[high_turn.index] / w.loc[high_turn.index].sum())).sum()
    # low_r = low_turn['chgPct'].mean()
    # high_r = high_turn['chgPct'].mean()
    return (b_date, low_r, high_r), (high_turn.index, low_turn.index)
```

```python
dates = [i for i in cal['calendarDate'] if i > '2006-01-01' and i < '2019-03-01' ]
turn_list = []
turn_dict = {}
for date in dates:
    val, tickers = get_turn_return_info(date) 
    turn_list.append(val)
    turn_dict[date] = tickers

res_turn_df = pd.DataFrame(turn_list, columns= ['date', 'low_return', 'high_return'])
res_turn_df.set_index('date', inplace = True)
```

```python
dates = [i for i in cal['calendarDate'] if i > '2006-01-01' and i < '2019-03-01' ]
cgo_list = []
cgo_dict = {}
for date in dates:
    val, tickers = get_mom_return_info(date)
    cgo_list.append(val)
    cgo_dict[date] = tickers

res_mom_df = pd.DataFrame(cgo_list, columns= ['date', 'low_return', 'high_return'])
res_mom_df.set_index('date', inplace = True)
```

we got the starage that we will buy the stocks which is neither lower cgo or higher turnover rate

```python
def return_of_starage(date, turn_dict = turn_dict, cgo_dict = cgo_dict):
    index_tickers = DataAPI.IdxCloseWeightGet(secID=u"",ticker=u"000905",beginDate= date , endDate= date, field=['consTickerSymbol', 'weight' ],pandas="1")
    index_tickers = index_tickers.set_index('consTickerSymbol')
    tickers = index_tickers.index.tolist()
    turn_tick_h,  turn_tick_l= turn_dict[date]
    cgo_tick_h, cgo_tick_l = cgo_dict[date]
    sub_tickers = [i for i in tickers if i not in turn_tick_h and i not in turn_tick_l and i not in cgo_tick_l and i not in cgo_tick_h]
    res = DataAPI.MktEqumGet(secID=u"",ticker=sub_tickers ,monthEndDate=u"",beginDate= date,
                                endDate= date, isOpen=u"1",field=['ticker', 'chgPct'],pandas="1")
    res.set_index('ticker', inplace = True)
    # weight = DataAPI.MktEqudGet(secID=u"",ticker= sub_tickers, tradeDate= e_date, beginDate=u"",
    #                             endDate=u"",isOpen="",field=['ticker', 'marketValue'],pandas="1")
    # weight.set_index('ticker', inplace = True)
    # w = weight['marketValue']
    w = index_tickers['weight']
    w = w.loc[res.index] / w.loc[res.index].sum()
    return (res['chgPct'] * w).sum()
```

```python
res_stag = pd.Series({ date:return_of_starage(date) for date in turn_dict.keys()})
```

```python
res_turn_df['strategy'] = res_stag
res_mom_df['strategy'] = res_stag
```
