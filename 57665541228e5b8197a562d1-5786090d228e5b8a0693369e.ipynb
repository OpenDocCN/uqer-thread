下面是帖子中原来的代码，我加了一些注释。

```python
from numpy import *
import pandas as pd

                           
used_factors = ['PE', 'NetProfitGrowRate', 'MA10', 'MA60', 'LCAP', 'LFLO', 'NPToTOR', 'OperatingProfitGrowRate', 'TotalAssetGrowRate', 'DHILO', 'DEGM', 'Skewness', 'DAREC', 'GREC']  #选择因子集
def get_stock(day):  
    universe = DataAPI.IdxConsGet(ticker="000300",intoDate=day,isNew="1",field=["consShortName","consID", "consTickerSymbol", "consExchangeCD"], pandas="1")
    stock=universe['consID'].values.tolist()
    stock_str = [str(x) for x in stock]
    return stock_str                     
#上面的步骤貌似可以用account.universe替代

def get_factor(day):
    stock_str=get_stock(day)       #account.universe
    factor =  DataAPI.MktStockFactorsOneDayGet(tradeDate=day,secID=stock_str,field=['secID','tradeDate']+used_factors,pandas="1") #取used_factors里的因子
    factor['ticker'] = factor['secID'].apply(lambda x: x[0:6]) 
    factor.set_index('ticker',inplace=True)
    for f in used_factors:  #遍历因子
        if factor[f].std() == 0:  #如果因子标准差等于0,下一个因子
            continue
        factor[f] = (factor[f] - factor[f].mean()) / factor[f].std()#如果因子不等于0，factor[yinzi]=（factoe[yinzi]-yinzi平均值）/标准差
    factor=factor.dropna() 
    # rank data
    factor.iloc[:,2:]=factor.iloc[:,2:].rank(method='first').apply(lambda x : x/len(factor)) #rank是排序，将值转化成顺序，这里是转化成从小到大之后，再除以总个数，详单与百分比
    return factor
#返回的是一个DF，所有因子都处理过了。




#以下两个函数式单层决策树生成函数
def stumpClassify(dataMatrix,dimen,threshVal,threshIneq):#just classify the data    dimen应该是datamatrix的某一列 名字
    #通过阈值比较返回-1或1
    retArray = ones((shape(dataMatrix)[0],1))            #建立一个datamatrix一样行数，1列的全是1的矩阵
    if threshIneq == 'lt':                               
        retArray[dataMatrix[:,dimen] <= threshVal] = -1.0
    else:
        retArray[dataMatrix[:,dimen] > threshVal] = -1.0
    return retArray
def buildStump(dataArr,classLabels,D):
    dataMatrix = mat(dataArr); labelMat = mat(classLabels).T
    m,n = shape(dataMatrix)
    numSteps = 10.0; bestStump = {}; bestClasEst = mat(zeros((m,1)))#这里的numsteps估计要修改，书上的范围是0-2，所有下面是从-1到10的遍历，
    minError = inf #init error sum, to +infinity      正无穷
    
    for i in range(n):#loop over all dimensions
        rangeMin = dataMatrix[:,i].min(); rangeMax = dataMatrix[:,i].max();
        stepSize = (rangeMax-rangeMin)/numSteps             #步长，最小到最大，这里是分10次，步长是0.1
        
        for j in range(-1,int(numSteps)+1):#loop over all range in current dimension  numsteps的作用是在特征的所有可能的值上进行遍历
            
            for inequal in ['lt', 'gt']: #go over less than and greater than
                threshVal = (rangeMin + float(j) * stepSize)     #最小值+第几次遍历*步长，也就是本次取值
                predictedVals = stumpClassify(dataMatrix,i,threshVal,inequal)#call stump classify with i, j, lessThan   利用stumpclassify把每一个因子i，因子值，大于或小于因子值。
                
                errArr = mat(ones((m,1)))
                errArr[predictedVals == labelMat] = 0
                weightedError = D.T*errArr  #calc total error multiplied by D
                #计算加权错误率
               
            
            
                if weightedError < minError: 
                    minError = weightedError#如果本次遍历的错误率低于最小错误率，那么记录这次
                    bestClasEst = predictedVals.copy()
                    bestStump['dim'] = i
                    bestStump['thresh'] = threshVal
                    bestStump['ineq'] = inequal
    return bestStump,minError,bestClasEst   #最后返回单层决策树返回的最佳参数
#以上两个函数式单层策略树生成函数



def adaBoostTrainDS(dataArr,classLabels,numIt=40):   #数据集，类别标签，numit=迭代次数，numit是整个adaboost算法中唯一需要用户设置的参数（训练几次）
    #DS的意思是( decision stump）单层决策树
    weakClassArr = []
    m = shape(dataArr)[0]
    D = mat(ones((m,1))/m)   #init D to all equal   D返回的是平均权重，1/m ，如果有10个
    aggClassEst = mat(zeros((m,1)))
    for i in range(numIt):
        bestStump,error,classEst = buildStump(dataArr,classLabels,D)#build Stump   获得本次遍历的单层决策最佳参数
        #print "D:",D.T
        alpha = float(0.5*log((1.0-error)/max(error,1e-16)))#calc alpha, throw in max(error,eps) to account for error=0 计算alpha，就是这次遍历的弱分类器权重
        bestStump['alpha'] = alpha  
        weakClassArr.append(bestStump)                  #store Stump Params in Array

        
        #为下一次迭代计算新的权重向量D（D权重是因子权重，alpha是分类器权重）
        expon = multiply(-1*alpha*mat(classLabels).T,classEst) #exponent for D calc, getting messy
        D = multiply(D,exp(expon))                              #Calc New D for next iteration
        D = D/D.sum()
        
        #错误率累加计算
        #calc training error of all classifiers, if this is 0 quit for loop early (use break)
        aggClassEst += alpha*classEst
        aggErrors = multiply(sign(aggClassEst) != mat(classLabels).T,ones((m,1)))
        errorRate = aggErrors.sum()/m
        
       
        if errorRate == 0.0: break   #如果错误率降低到0，直接退出训练
    return weakClassArr

def adaClassify(datToClass,classifierArr):                          #利用训练处的多个弱分类器进行分类，该函数输入的是由一个或多个待分类样例dattoclass以及多个若分类器组成的数组classifierarr
    dataMatrix = mat(datToClass)#do stuff similar to last aggClassEst in adaBoostTrainDS
    m = shape(dataMatrix)[0]             #取有几行，有几个股票
    aggClassEst = mat(zeros((m,1)))      #建立一个m行1列的0矩阵
    for i in range(len(classifierArr)):      #遍历classifierarr中的若分类器
        classEst = stumpClassify(dataMatrix,classifierArr[i]['dim'],classifierArr[i]['thresh'],classifierArr[i]['ineq'])#call stump classify  基于stumpclassify对每个分类器得到的一个类别的估计值
        aggClassEst += classifierArr[i]['alpha']*classEst    #类别估计值乘上该单层决策树的alpha权重然后累加到aggclassest上。
        #print aggClassEst
    return aggClassEst  #返回的是aggclassest的符号，如果大于0，返回1，小于0返回-1
```
