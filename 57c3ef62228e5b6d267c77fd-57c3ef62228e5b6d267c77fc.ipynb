### 说明：策略所交易的标的为50ETF期权本身，而不是根据期权的成交数据来择时交易股票

而且由于没有期权策略的回测，所以所有的交易账户统计等只能自己写啦
策略思想参考了方正证券金融工程研究报告《20160506-方正证券-期权时间价值专题研究：一寸光阴一寸金，期权可买寸光阴》




策略组合：
 -  卖出当月平值认购期权
 -	卖出当月平值认沽期权
 -	买入下月平值认购期权
 -	买入下月平值认沽期权
   （相当于2个日历差价合约组合，或者2个跨式合约组合)

移仓频率：
 -	每个月换仓一次。当月合约到期日，调整合约，所有合约平仓，在第二天建立新的当月和下月合约组合
    
组合调整：
 -	换仓条件：ETF收盘价超过现持有的平值合约行权价格，换仓为最接近ETF收盘价的平值合约

回测时间:
 -	时间区间从2015年2月9日到2016年8月29日
