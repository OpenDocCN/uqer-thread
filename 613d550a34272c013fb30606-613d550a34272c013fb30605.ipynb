# 基金行业归因

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt


beginDate='20160101'
endDate='20161231'

#长信医疗保健163001
df1=DataAPI.MktFunddGet(secID=u"",ticker=u"163001",tradeDate=u"",beginDate=beginDate,endDate=endDate,field=u"tradeDate,closePrice,preClosePrice",pandas="1")
#沪深300指数399300
df2=DataAPI.MktIdxdGet(indexID=u"",ticker=u"399300",tradeDate=u"",beginDate=beginDate,endDate=endDate,exchangeCD=u"XSHE,XSHG",field=u"tradeDate,closeIndex,preCloseIndex",pandas="1")
#Shibor3M
df3=DataAPI.MktIborGet(secID=u"",ticker=u"Shibor3M",tradeDate=u"",beginDate=beginDate,endDate=endDate,currency=u"",field=u"tradeDate,rate",pandas="1")

#处理rf和其他数据日期不一致问题
df=pd.merge(df1,df2,how='inner',on='tradeDate')
merge=pd.merge(df,df3,how='inner',on='tradeDate')

r_fund=np.log(np.array(merge['closePrice'])/np.array(merge['preClosePrice']))     #基金收益率序列
r_f=np.array(merge['rate'])/100       #无风险利率Shibor3M

#申万一级行业指数
sw_indus=DataAPI.IndustryGet(industryVersion=u"",industryVersionCD=u"010303",industryLevel=u"1",isNew=u"1",prntIndustryID=u"",industryName="",industryID=u"",field=u"indexSymbol,industryName",pandas="1")
#indus=DataAPI.MktIdxdGet(indexID=u"",ticker=u'801210',tradeDate=u"",beginDate=u"20160101",endDate=u"20161231",exchangeCD=u"XSHE,XSHG",field=u"tradeDate,closeIndex,preCloseIndex",pandas="1")

list_r2,list_beta,list_tr=[],[],[]
for code in sw_indus['indexSymbol']:
    indus=DataAPI.MktIdxdGet(indexID=u"",ticker=code,tradeDate=u"",beginDate=beginDate,endDate=endDate,exchangeCD=u"XSHE,XSHG",field=u"tradeDate,closeIndex,preCloseIndex",pandas="1")
    r_indus=np.log(np.array(indus['closeIndex'])/np.array(indus['preCloseIndex']))
    x,y=r_indus-r_f,r_fund-r_f
    x=x.reshape(len(x),1)
    c=np.ones((len(x),1))
    X=np.hstack((c,x))
    res=(sm.OLS(y,X)).fit()
    list_r2.append(res.rsquared)
    list_beta.append(res.params[1])
    list_tr.append(np.log(indus['closeIndex'][indus['closeIndex'].argmax()]/indus['closeIndex'][0])-r_f.mean())
res_indus=pd.DataFrame([])
res_indus['指数代码']=sw_indus['indexSymbol']
res_indus['指数名称']=sw_indus['industryName']
res_indus['拟合R方']= list_r2
res_indus['beta']= list_beta
res_indus['期间总收益']= list_tr
res_indus['开始时间']= beginDate
res_indus['终止时间']= endDate
res_indus=res_indus.sort_values('拟合R方',ascending=False)
res_indus.head()

```
