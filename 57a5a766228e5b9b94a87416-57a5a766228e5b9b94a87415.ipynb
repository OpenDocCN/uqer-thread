```python
import scipy.fftpack as fftpack
import numpy as np
import pandas as pd
import datetime
from CAL.PyCAL import *
import matplotlib.pyplot as plt
```

```python
def hb(x):                      ####希尔伯特变换
    if x>0:
        a = 1
    else:
        a = 0
    return a
```

```python
def time(x = nowtime ):               ########希尔伯特交易函数
    begin = cal.advanceDate(x ,'-1850B',BizDayConvention.Following).strftime('%Y%m%d')
    a = DataAPI.MktIdxdGet(tradeDate=u"",indexID=u"",ticker=u"000002",beginDate= begin ,endDate= x ,field=u"",pandas="1")
    a['ma'] = pd.rolling_mean(a['preCloseIndex'],len(a),min_periods=20)
    a['diff'] = a['ma'].diff()
    a = a.dropna()
    a['hilbert'] = fftpack.hilbert(a['diff'])
    a = a.dropna()
    a['hb'] = a['hilbert'].apply(hb)
    return a.hb[-1:]
```

```python
from scipy import fftpack
import numpy as np
import matplotlib.pyplot as pl

# 产生1024点4个周期的正弦波
t = np.linspace(0, 4*np.pi, 1024, endpoint=False)
x = np.sin(t)

# 进行Hilbert变换
y = fftpack.hilbert(x)
pl.plot(x, label=u"原始波形")
pl.plot(y, label=u"Hilbert转换后的波形")
pl.legend()
pl.show()
```
