# 获取沪深300股息率最高的十只股票

```python
from __future__ import division
import pandas as pd
import numpy as np
import datetime
from dateutil.parser import parse
from CAL.PyCAL import *
# import os
cal = Calendar('China.SSE')
```

```python
trade_date = '20190816'

# 获取成分股及权重
# stocks = DataAPI.IdxCloseWeightGet(secID=u"",ticker=u"000300",beginDate=u"20190101",endDate=u"20190901",field=u"consID,consShortName",pandas="1")
stocks = DataAPI.IdxConsGet(secID=u"",ticker=u"000300",isNew=u"1",intoDate=trade_date,field=u"",pandas="1")

universe = stocks["consID"].tolist()
#stocks.head()
#去重
universe=list(set(universe))
#universe
#获取股息率
df_factor = DataAPI.MktStockFactorsOneDayGet(secID=universe,tradeDate=trade_date,field=['secID','CTOP'],pandas="1")
df_factor.sort_values('CTOP', ascending=False, inplace=True)
# 选取最高的10家
df_factor_select = df_factor.head(20)
df_factor_select["secID"]=df_factor_select.secID.map(stocks.set_index('consID').consShortName.to_dict())

df_factor_select
```
