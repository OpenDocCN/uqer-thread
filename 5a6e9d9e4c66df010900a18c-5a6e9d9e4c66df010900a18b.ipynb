# 2. 指标介绍

---

- Aroon是一个用于识别资产价格是否存在趋势以及衡量趋势强度的技术指标，与一般关注相对于时间的价格的趋势指标的区别在于，Aroon关注的是相对于价格的时间
- Aroon一般分为两个指标，Aroon-Up和Aroon-Down，分别记录了距离最近一段时间的最高/低价的时间
- 计算公式如下
	- Aroon-Up = ((T - Days Since T-day High)/T) x 100
	- Aroon-Down = ((T - Days Since T-day Low)/T) x 100
	- T一般取25
