```python
# 以下是画了一下 大证指数和 招商银行从2014年8月以下的走势图，从图中可看到，上证从5000点下跌过程中，招行多次摆脱大盘上涨(原因你懂的)
import datetime as dt
import numpy as np
import seaborn as sns
sns.set_style('white')
from matplotlib import pyplot as plt
font.set_size(20)

zicn = '000001.ZICN'
zsyh = '600016.XSHG'
beginDate = '20140830'
endDate = '20160420'

data = DataAPI.MktIdxdGet(indexID=zicn, beginDate=beginDate, endDate=endDate)
data.index = data.tradeDate.apply(lambda x: dt.datetime.strptime(x, '%Y-%m-%d'))
data['zicn'] = data.closeIndex
data = data[['zicn']]

data2 = DataAPI.MktEqudAdjAfGet(secID=zsyh, beginDate=beginDate, endDate=endDate)
data2.index = data2.tradeDate.apply(lambda x: dt.datetime.strptime(x, '%Y-%m-%d'))
data['zsyh'] = data2.closePrice * 50

data.plot(figsize=(8,4), grid = False)
plt.legend([u'上证指数', u'招商银行'], prop = font, loc = 'best')
sns.despine()
```
