```python
import numpy as np
import pandas as pd
from pandas import DataFrame,Series

from matplotlib import pyplot as plt
```

```python
#回测区间
start = '2012-08-01'                       # 回测起始时间
end   = '2015-08-01'                         # 回测结束时间

universe =[ '000001.XSHE']
```

以下研究获取哪种数据更可靠！

```python
#获取回测期间全部数据,DataAPI.MktEqudGet(tradeDate=u"20150513",secID=u"",ticker=u"",beginDate=u"",endDate=u"",isOpen="",field=u"",pandas="1")
#获取沪深AB股日行情信息，包含开盘价、最高价、最低价、收盘价、成交量等字段，每日16:00更新
df = DataFrame()
df = DataAPI.MktEqudGet(secID=universe[0],beginDate=start,endDate=end,field=u"tradeDate,closePrice",pandas="1")
df = df.rename(columns={'tradeDate':'Date','closePrice':'Close'})
df['Date'] =  pd.to_datetime(df['Date']) #因为DataAPI所有日期都是以字符串形式存储的，因此此处需要转换成日期时间型,这里很重要！
df = df.set_index('Date')#转换后index是DatetimeIndex型
df.head(5)
```

### 这是行情数据

```python
df_Adj = DataFrame()
df_Adj = DataAPI.MktEqudAdjGet(secID=universe[0],beginDate=start,endDate=end,field=u"tradeDate,closePrice",pandas="1")
df_Adj = df_Adj.rename(columns={'tradeDate':'Date','closePrice':'AdjClose'})
df_Adj['Date'] =  pd.to_datetime(df_Adj['Date']) #因为DataAPI所有日期都是以字符串形式存储的，因此此处需要转换成日期时间型,这里很重要！
df_Adj = df_Adj.set_index('Date')#转换后index是DatetimeIndex型
df_Adj.head(5)
```

### 这是DataAPI.MktEqudAdjGet提供的前复权数据，站在今天收盘后看到的前复权数据

```python
df0 = pd.concat([df, df_Adj], axis=1)
df0.head()
```

```python
#获取获取沪深A股和B股用来调整行情的前复权因子数据，包含除权除息日、除权除息事项具体数据、本次复权因子、累积复权因子以及因子调整的截止日期。该因子用来调整历史行情，不作为预测使用，于除权除息日进行计算调整。
adj_Factor = DataAPI.MktAdjfGet(secID=universe[0],beginDate=start,endDate=end,field=u"endDate,accumAdjFactor",pandas="1")
adj_Factor = adj_Factor.rename(columns={'endDate':'Date'})
adj_Factor['Date'] =  pd.to_datetime(adj_Factor['Date']) #因为DataAPI所有日期都是以字符串形式存储的，因此此处需要转换成日期时间型,这里很重要！
adj_Factor = adj_Factor.set_index('Date')#转换后index是DatetimeIndex型
adj_Factor.head()
```

### 这是累积复权因子以及因子调整的截止日期

```python
df1 = pd.concat([df0, adj_Factor], axis=1)
#处理NaN
df1 = df1.fillna(method='bfill')  #  在列方向上以后一个值作为值赋给NaN
df1 = df1.fillna(method='ffill')  #  在列方向上以前一个值作为值赋给NaN
df1['adjClose'] = df1['Close']*df1['accumAdjFactor']
df1.loc['2012-10-17':'2012-10-25']
```

#### adjClose是站在历史某一天的看的前复权数据，可见与AdjClose差别很大

## 下面以两种复权数据来计算策略的夏普比率

纯多头策略:2012年8月1日以收盘价买入一股中国平安（000001.XSHE）并持有，2015年8月1日以当日收盘价卖出平仓。假设持股期内无风险利率是每年4%。

### 第一种：DataAPI.MktEqudAdjGet 提供的前复权数据

```python
df = df1.iloc[1:][['AdjClose']]
df.head()
```

```python
#日度收益
df['returns'] = df.pct_change()
df.fillna(0, inplace = True)  #  将NaN置换为0
df.head()
```

```python
#超额收益
risk_free_rate = 0.04  #年无风险利率
df['Excess_Dailyret'] = df['returns'] - risk_free_rate/250
df.head()
```

```python
#夏普率公式
def create_sharpe_ratio(returns, periods=250):
    """
    Create the Sharpe ratio for the strategy, based on a 
    benchmark of zero (i.e. no risk-free rate information).

    Parameters:
    returns - A pandas Series representing period percentage returns.
    periods - Daily (250), Hourly (250*4), Minutely(250*4*60) etc.
    """
    return np.sqrt(periods) * (np.mean(returns)) / np.std(returns)
```

```python
#日度夏普率
daily_sr =create_sharpe_ratio(df['Excess_Dailyret'])
daily_sr  #这个结果是购买并持有策略的夏普比。
```

```python
def plot_returns(returns, title=None):
    returns.plot(figsize=(12,8))
    plt.xlabel('Year')
    plt.ylabel('PnL')
    if title is not None: plt.title(title)
    plt.show()
```

```python
nav = (1+df['Excess_Dailyret']).cumprod()
plot_returns(nav)
```

### 第二种：DataAPI.MktEqudAdjGet 提供的前复权数据

```python
df = df1.iloc[1:][['adjClose']]
df.head()
```

```python
#日度收益
df['returns'] = df.pct_change()
df.fillna(0, inplace = True)  #  将NaN置换为0
df.head()
```

```python
#超额收益
risk_free_rate = 0.04  #年无风险利率
df['Excess_Dailyret'] = df['returns'] - risk_free_rate/250
df.head()
```

```python
#日度夏普率
daily_sr =create_sharpe_ratio(df['Excess_Dailyret'])
daily_sr  #这个结果是购买并持有策略的夏普比。
```

```python
nav = (1+df['Excess_Dailyret']).cumprod()
plot_returns(nav)
```
