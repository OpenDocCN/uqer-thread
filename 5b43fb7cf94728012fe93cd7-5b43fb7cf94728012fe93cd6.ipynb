# 个股财务和估值摘要

查个股的核心指标，计算出个股估值的历史百分位和各指标的行业内百分位排名

```python
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import datetime
```

```python
#获取个股当前各指标信息，delay表示要不要取昨天的数据 
def stkDescribe(tiker='000001',delay=True):    
    fields=['ticker','tradeDate','PB','PE','ROE','ROE5','ROA','GrossIncomeRatio','NetProfitGrowRate','OperatingRevenueGrowRate','DebtsAssetRatio']
    #查个股指标和历史百分位
    if delay:
        end = datetime.datetime.now().date()-datetime.timedelta(1)
    else:
        end = datetime.datetime.now().date()
    facs = DataAPI.MktStockFactorsDateRangeGet(secID=u"",ticker=tiker,beginDate=u"20130101",endDate=end,field=fields,pandas="1")
    curfacs = facs[-1:]
    PB_p = round(stats.percentileofscore(facs.PB.values,curfacs.PB.values[0]),2)
    PE_p = round(stats.percentileofscore(facs.PE.values,curfacs.PE.values[0]),2)
    print '股票',tiker,'的关键指标：'
    print '估值'
    print 'PB值:',curfacs.PB.values[0],'历史百分位:',PB_p
    print 'PE值:',curfacs.PE.values[0],'历史百分位:',PE_p
    print '盈利'
    print 'ROE:',curfacs.ROE.values[0]*100,'%'
    print '5年ROE:',curfacs.ROE5.values[0]*100,'%'
    print 'ROA:',curfacs.ROA.values[0]*100,'%'
    print '毛利润率:',curfacs.GrossIncomeRatio.values[0]*100,'%'
    print '成长'
    print '净利润增长率:',curfacs.NetProfitGrowRate.values[0]*100,'%'
    print '营业收入增长率:',curfacs.OperatingRevenueGrowRate.values[0]*100,'%'
    print '安全'
    print '资产负债率:',curfacs.DebtsAssetRatio.values[0]*100,'%'    
    print '---------------------------------'
    #查个股行业代码
    induIDdf = DataAPI.EquIndustryGet(ticker=tiker,industryVersionCD=u"010303",field=u"industryID2,industryName2,isNew",pandas="1")
    induIDdf=induIDdf[induIDdf.isNew==1]
    print '所属申万二级行业：',induIDdf.industryName2.values[0]
    #查行业个股集合
    stksInIndu = DataAPI.EquIndustryGet(industryVersionCD=u"010303",industryID2=induIDdf.industryID2.values[0],field=u"ticker,isNew",pandas="1")
    stksInIndu = stksInIndu[stksInIndu.isNew==1][['ticker']]
    stks = list(stksInIndu.ticker)
    #行业均值
    stksData = DataAPI.MktStockFactorsOneDayGet(tradeDate=curfacs.tradeDate.values[0],secID=u"",ticker=stks,field=fields,pandas="1")
    m1 = stksData.median().map(lambda v:round(v,3))
    print '当前行业中位数：'
    print 'PB:',m1.PB
    print 'PE:',m1.PE
    print 'ROE:',m1.ROE*100,'%'
    print '5年ROE:',m1.ROE5*100,'%'
    print 'ROA:',m1.ROA*100,'%'
    print '毛利润率:',m1.GrossIncomeRatio*100,'%'
    print '净利润增长率:',m1.NetProfitGrowRate*100,'%'
    print '营收增长率:',m1.OperatingRevenueGrowRate*100,'%'
    print '资产负债率:',m1.DebtsAssetRatio*100,'%'
    #个股在行业中的排名
    stksData.index = stksData.ticker
    rankFileds=['PB','PE','ROE','ROE5','ROA','GrossIncomeRatio','NetProfitGrowRate','OperatingRevenueGrowRate','DebtsAssetRatio']
    per = stksData[rankFileds].apply(lambda s:int(stats.percentileofscore(s.values,s[tiker])))
    print '-----------------------------------'
    print '个股指标在行业中百分位：'
    print 'PB:',per.PB
    print 'PE:',per.PE
    print 'ROE:',per.ROE
    print '5年ROE:',per.ROE5
    print 'ROA:',per.ROA
    print '毛利润率:',per.GrossIncomeRatio
    print '净利润增长率:',per.NetProfitGrowRate
    print '营收增长率:',per.OperatingRevenueGrowRate
    print '资产负债率:',per.DebtsAssetRatio 
    #p=per[rankFileds]
    #p.plot(kind='bar')
stkDescribe()    
```
