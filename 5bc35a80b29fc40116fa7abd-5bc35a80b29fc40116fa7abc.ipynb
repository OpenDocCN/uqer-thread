# WeChat-pandas的groupby分组函数

```python
import numpy as np 
import pandas as pd
df=pd.DataFrame({'key1':list('aabbcaabbc'),'key2':list('AAAAABBBBB'),\
'data1':np.random.randint(0,20,10),'data2':np.random.randint(10,30,10)})
df
```

单指标分组

```python
#按照key1进行分组
grouped = df.groupby(['key1']) 
#查看分组后包含哪些属性和方法
dir(grouped) 
```

```python
# 展示分组后的数值型变量的描述性统计
grouped.describe()
```

```python
#展示按照key1分组后的第一个数，返回后面的数用grouped.tail(1)
grouped.head(1)
```

```python
# 利用agg/aggregate进行聚合
grouped.agg('min')
```

```python
# 指定聚合的列和聚合方法
grouped.agg({'data1': ['min', 'max'], 'data2': 'sum'})
```

```python
# 返回每一个分组的索引位置
grouped.groups
```

```python
#利用apply函数对表中数据乘以2
grouped.get_group('a').apply(lambda x:2*x)
```

```python
#利用apply函数对data1和data2进行归一化
grouped.get_group('a')[['data1', 'data2']]\
.apply(lambda x: (x - x.mean()) / x.std())
```

多指标分组

```python
# 将df按照key1和key2联合排序
df1 = df.sort(['key2', 'key1'])
df1
```

```python
#按照key2和key1联合分组
grouped1 = df1.groupby(['key2', 'key1'])
grouped1.describe()
```

```python
#返回A分组里面a子分组的值
grouped1.get_group(('A', 'a'))
```

```python
#返回多分组里面的第一行值
grouped1.head(1)
```
