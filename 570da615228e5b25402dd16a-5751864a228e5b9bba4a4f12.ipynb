```python
# Import libraries to find linear trend and plot data
import statsmodels.api as sm
import scipy.stats as stats
from scipy import stats
import scipy.spatial.distance as distance
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from CAL.PyCAL import * 
start = '20140101'
end = '20141230'
asset=DataAPI.MktEqudGet(ticker='000002',beginDate=start,endDate=end,field=['tradeDate','closePrice'],pandas="1").set_index('tradeDate')['closePrice'] # 使用Dataapi获取万科A的收盘价

# 绘制收盘价变化曲线
dates = asset.index

asset.plot(figsize=(12,8))
# 对收盘价数据进行直线拟合
X = np.arange(len(asset))
x = sm.add_constant(X)
model = stats.linregress(asset, x).fit() 
a = model.params[0]
b = model.params[1]
Y_fit = X * b + a
plt.plot(X, Y_fit, 'r', alpha=0.9);
plt.ylabel('Price')
plt.legend(['WKA', 'Trendline']);
```
