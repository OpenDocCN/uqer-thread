# 如何获取期货合约代码？
在写多因子对冲策略时，需要买入当前主力合约，但是并没有在优矿相应文档中看到相关合约代码的说明，希望这个帖子能帮助到大家

## 主力合约标识 示例
'IFM0'： 沪深300 主力合约


期货当月连续合约：

当月连续合约与主力连续合约的代号相似，只需要将品种代号后面拼上’L0’，‘L1’，‘L3’，‘L6’即可，如’IFL0’，‘IFL1’，‘IFL3’，'IFL6’表示沪深300期货当月、次月、当季、次季连续合约，如’CUL0’表示铜期货当月连续合约。

股指期货：标的代码为合约可选值和类型可选值的组合，其中：
合约可选值：‘IF’,‘IH’,‘IC’
‘L0’: 当月连续合约
‘L1’: 次月连续合约
‘L3’: 当季连续合约
‘L6’: 次季连续合约

构建方法如：
‘IFL0’, 沪深300股指期货当月连续合约
‘IHL1’, 上证50股指期货次月连续合约。
