3.数据集划分为训练集、验证集、测试集的意义？

* 训练集：用于模型训练的历史数据，我们是可以翻看这些样本，去做特征设计、错误分析的。

* 验证集：用于对比两个不同模型的好坏，提早发现过拟合、用于参数调优。不建议模型开发人员去翻看这些样本，可能会导致验证集作用失效。

* 测试集：可以理解为线上的真实数据，我们并不可见的数据集。用于最终评价模型表现。

5.机器学习模型分类

- 按大家关注的热点，可以分为
-- 经典机器学习模型，模型复杂度不高，挖掘历史规律。(Logistic Regression, Support Vector Machine, Gradient Boosting Decision Tree, Conditional Random Field, etc.)
-- 深度学习，模型复杂度高/参数多，挖掘历史规律。(Convolutional Neural Network, Recurrent Neural Network, Long Short-Term Memory, Deep Residual Network, Generative Adversarial Nets, Attention Based Model, etc.)
-- 增强学习，智能体可以根据环境、状态自助决策行为。(Q-Learning, Deep Q Network, Monte-Carlo Tree Search(Deep Policy/Value Network) etc.)

- 按目标/标签取值类型，可以分为
-- 分类，标签取值为离散值，进一步分为二分类、多分类。(多分类问题，可视为多个二分类问题，OneVsAll/Rest；也可对多个类别进行建模，最后用softmax)
-- 回归，标签取值为连续值。(Support Vector Regression, Oridinary Least Squares, Weighted Least Squares, Iteratively Reweighted Least Squares, etc.)

- 按是否需要提供目标/标签，可以分为
-- 监督学习，需要提供特征(X)及标签(Y)，可用于给出新样本所属类别的概率。(k-Nearest Neighbors, Linear Discriminant Analysis, etc.)
-- 无监督学习，仅需要提供特征(X)，可用于降维、提取共性特征、关联规则挖掘。(K-Means, Principal Components Analysis, Associate Rules, Word2vec/Doc2Vec, Latent Dirichlet Allocation, etc.)

- 按目标函数的建模方式，可以分为
-- 统计学派，目标函数中不包含概率形式。(大部分机器学习模型，既可以用统计学派建模，也可以用贝叶斯学派建模)
-- 贝叶斯学派，会引入先验概率、后验概率，对"不确定性"进行建模；再辅以Maximum Likelihood Estimation/Laplace Approximation/Sampling技术求解。根据是否对联合概率建模，可进一步分为判别式模型、生成式模型。缺点是，先验概率分布，只能在少数存在共轭分布的分布中进行选择。(Naive Bayes, Gaussian Process, Probabilistic Graphical Model, Gaussian Mixture Model, Kernel Density Estimator, etc.)

- 按应用场景，可以分为
-- 排序学习/点击率估计/CTR，根据用户历史行为，对被召回的样本，给出预测点击概率。(LambdaRank, ListNet, etc.)
-- 个性化推荐。(SVD++, Factorization Machines, Wide&Deep Learning, etc.)
-- 自然语言处理，包括序列标注、机器翻译、文本分类、自动摘要等子领域。(Encoder-Decoder, Seq2Seq, etc.)
-- 计算机视觉，包括显著性/物体检测、光流场、目标追踪、ReID、目标识别、图像分割、场景理解、场景重建等子领域。(Fisher/Riemannian Metric, Spectral Clustering, AdaBoost, etc.)

# <font color=#0099ff>**第一部分：机器学习核心三要素**</font>

上面我们介绍了一票机器学习算法，如果之前没有深入了解，听起来肯定一头雾水。没关系，为了方便大家理解，我们借鉴学术圈的思想，认为所有的机器学习算法都包括，表示/目标函数、评估/距离度量准则、优化三要素。大家学习某一机器学习算法时，都可以试着用这种框架去理解。本节会介绍以下方面

- 目标函数、距离度量准则、优化方法

- 各模型用三要素描述的概览图

1.机器学习算法(Learning)=表示(Representation)+评估(Evaluation)+优化(Optimization)

- 表示，机器学习问题可以用函数形式来表达，不同模型的建模方式会有区别，比如样本点、决策树、超平面、图模型等。

- 评估，旨在解决“如何评估”同一模型在不同参数条件下的优劣，比如hinge loss, cross entropy loss, square loss等。

- 优化，给定训练样本、优化方程后，如何求解模型中的参数，一般使用数值优化、线性规划、非线性规划、凸优化、离散优化等学科的算法。比如branch and bound, GD, SGD, Adagrad, Adam等。


2.<font color=red>我们用上述三要素，拆解linear regression</font>

- 以多因子模型为例，我们认为个股收益为标签$y_{i}=r_{i}$，可以被$x_{i}$(包括，风格暴露$x_{i}^{style}$、行业暴露$x_{i}^{industry}$)所解释。

- 表示，目标函数为$g_{w}$，$min_{w^{\*}} g_{w} = min_{w^{\*}} \frac{1}{N} \sum_{i=1}^{N} (w^T x_{i} - y_{i})^2$

- 评估，如目标函数所示，我们采用的是square loss(误差越大在目标函数中的影响就会越大，2次型便于求解)，误差即为特质性收益$\epsilon_{i}=w^{\*T} x_{i} - y_{i}$

- 优化，在不加其他约束条件下，是一个凸问题，梯度为0的极值点即为所求。梯度为$\nabla g_{w} = \frac{2}{N}(x^T x w - x^T y) = 0$；在$N \gg d+1$前提下，最优解$w_{\*}=(x^T x)^{-1} x^T y$；其中$N$为样本数，$d$为特征维度

- 进一步，如果我们对目标函数，添加模型复杂度的l1-norm约束，优化问题变为下式。那么问题不再有解析解，可以使用内点法等迭代算法求解。
    $$ min_{w^{\*}} \frac{1}{N} \sum_{i=1}^{N} (w^T x_{i} - y_{i})^2$$
    $$s.t. ||w||_{1} \le C$$

# <font color=#0099ff>**第二部分：如何评估模型好坏**</font>

当我们通过上步，训练得到了一个机器学习模型，怎么使用测试集去评估就成为了重点。通过本节学习，大家可以应对实际场景中遇到的问题，比如"我的模型还有进一步提升空间吗"、"训练集准确率很高，但在测试集中表现很差"、"准确率很高、但召回率很低，这样的模型有用吗"、"模型复杂度越高越好吗"等。我们会着重看以下方面

- bias/variance

- 模型复杂度/curse of dimension

- 准确/召回/F1/AUC

1.机器学习误差=偏差(bias)+方差(variance)
- 偏差，指的是模型在训练集上的最终准确率，可以用$E_{in}$表示；偏差=不可避免误差(人类识别的水平)+可避免的误差
-- 不可避免误差，一般可以用人类识别的平均水平来代替。因为如果人都无法提炼经验、规律，从而自主识别出来，机器就更难做到这点。
-- 可避免的误差，指的是机器相较于人表现差的部分，即可以进一步提升空间。

- 方差，指的是模型在测试集与训练集之间的表现差异。方差=模型在测试集上的准确率($E_{out}$)-模型在训练集上的准确率(偏差/$E_{in}$)；$E_{out}$表示测试集上的准确率

- 我们希望模型具有泛化(Generalization)能力，即$E_{in} \approx E_{out}$，但往往不会这么理想

- 首先可以根据实际问题，设定一个可接受误差$\delta$，代表这个模型可以线上使用。那么根据$E_{in}, E_{out}, \delta$三者的不同组合情况，原因分析及解决方案也不同:
-- $E_{out} > E_{in} > \delta$，可能问题本身很复杂，难以达到可用水平$\delta$；也可能模型在训练集中，由于训练数据过少、模型太简单、特征不具有区分度等原因，导致<font color=red>欠拟合</font>。
-- $E_{out} \gg \delta > E_{in}$，可能由于模型参数过多/太复杂、训练数据过少，导致<font color=red>过拟合</font>。
-- $E_{out} > \delta > E_{in}$，这是一种理想情况，如下图，模型精度随着数据集增加，明显收敛。模型在训练集上的表现$E_{in}$小于$\delta$；模型在测试集上的表现$E_{out}$也比较接近$\delta$。

- 若具有较高的可避免偏差，可以尝试增加模型复杂度、<font color=red>进行错误分析(修改、添加特征)</font>。

- 若具有较高的方差，可以尝试增加训练数据集、正则化。

2.模型复杂度/VC Dimension/Curse of Dimension/Occam's Razor
- 模型复杂度，一般受模型中参数个数、模型本身影响。比如用线性回归模型，去解决线性不可分的数据集，发现效果不好；改用多项式回归模型，从而增加了模型复杂度。

- VC维(Vapnik-Chervonenkis Dimension)，解释并回答了"机器学习模型能否从给定数据集中学习到规律"这一问题。VC维指，能被假设空间$H$，所shatter的最大数据集大小n。

- 维数灾难(Curse of Dimension)，当特征维度上升，使得每维特征上的样本点变得稀疏；若需要保证每维特征的覆盖，所需样本的数量会指数级上升，所以模型中特征也并不是越多越好。

- 奥卡姆剃刀(Occam's Razor)，也指模型复杂度与实际问题要适配，不要盲目追求复杂模型。

- 如下图，当模型复杂度提高后，我们发现训练集误差$E_{in}$在降低，但样本外误差$E_{out}$会变大，泛化能力反而变差，即出现了常说的过拟合现象。

3.评估指标
- 准确率，指模型预测为正样本，其中实际也是正样本的比例。$Precision=\frac{TP}{TP+FP}$

- 召回率，指模型预测且实际均为正样本，占实际为正样本的比例。$Recall=\frac{TP}{TP+FN}$

- F1值，是更为全面的指标，既考虑准确率、又考虑召回。$F1=2\cdot \frac{Precision \cdot Recall}{Precision + Recall}$

- AUC(Area Under the Curve)/ROC(Receiver operating characteristic)。由于错检率(FPR)与漏检率(TPR)之间存在tradeoff，AUC指标更为合理。因为机器学习模型，可以预测样本属于正样本的概率$p$；我们通过调节阈值$thres$，来决定是正样本($p \ge thres$)，还是负样本($p \lt thres$)；即可计算多个(FPR,TPR)点，从而画出ROC曲线

- 混淆矩阵(Confusion Matrix)，用于多分类问题，可以更清晰的看出，究竟模型在哪一类别上表现不好。

- 不同实际问题，都会有自己独特的评估方法；<font color=red>建议动手搭建模型前，先花2~3天决定评价方案</font>。比如，针对搜索排序问题，衡量排序结果的质量，可以使用NDCG指标等。

# <font color=#0099ff>**第三部分：经典机器学习中常见的技术**</font>

我们在第一部分，学习了机器学习算法三要素。在此基础上，工业界、学术界还有一些常见技术，用于提高模型精度、适用性、复杂度、鲁棒性等。我们接下来逐一介绍：

- 常见特征处理方法

- 特征选择

- 交叉验证

- 正则化

- 核化

- 集成学习

1.常见特征处理方法，包括特征的预处理、特征组合、特征空间映射，是提高模型表现的切实一环。
- 离散化，用于将文本特征进行数值化表示，便于计算。比如个股所属行业(假设申万28个行业)，可以变为$1x28$维向量，仅在某一维度取值为1，其余为0。

- 归一化，使得不同特征维度在量级上可比，便于计算、优化收敛。比如个股营业收入、PE两个特征，原本是不同量级；我们可以用优矿中的winsorize/standardize函数来实现归一化，从而在量级上一致。

- 组合特征，可以增加特征维度，增强模型表示能力。比如宏观变量(有通胀、通缩、滞涨、紧缩四个状态)、个股行业情况(有幼稚期、成长期、成熟期、衰退期四个状态)两维特征，通过做笛卡尔积，得到16维0/1特征。

- 降维，当我们面临特征数量巨大情况时，可以通过PCA等方式将原始特征映射到隐变量空间，实现降维的目的。在计算机视觉领域，往往能取得较好的效果。

- 使用GBDT的叶子结点的输出值，作为新的特征，即实现了特征空间的映射。GBDT模型的基础是决策树，最终每个叶子结点会给出属于正样本的概率。例如，Wide&Deep Learning算法。

2.特征选择，当机器学习模型面对海量特征时，如何筛选特征、降低模型复杂度、目标函数的搜索空间，都是重要课题。
- filter，利用特征与目标标签之间的关系，筛选重要特征。包括<font color=red>相关系数(有点像因子分析中的IC检验)</font>、卡方检验、信息增益、互信息等。

- wrapper，通过目标函数的AUC/MSE，来决定是否加入一维新特征。包括完全搜索、启发式搜索等。

- embedded，目标函数建模中，考虑如何筛选特征，优化求解同时自动筛选特征。包括L1范数、L2范数、决策树中的熵/信息增益、深度学习的隐藏层等。

3.交叉验证(Cross Validation)，是为了在开发数据中，使用<font color=red>训练集</font>训练模型，再使用<font color=red>验证集</font>来评估模型好坏，进行调优；避免到线上的测试集才发现问题。
- <font color=red>第一步</font>，划分训练集、验证集pairs

- <font color=red>第二步</font>，分别在各个pairs上进行训练、测试，取平均值作为模型表现，进行评估

- K-fold，原始数据分为开发集、测试集；开发集再分为K等份，其中K-1份作为训练集，剩余1份作为验证集；得到K对pairs。

- Leave One Out，是K-fold(K=N，其中N为样本数)时的特例；得到N对pairs。

- Leave P Out，将原始数据，按$C_{N}^{P}$进行组合排列，得到训练集、验证集的pairs。

4.正则化(Regularization)，是为了防止模型复杂度过高(参数过多)，导致在训练集中过拟合，而在测试集中表现过差，通过在目标函数中添加下面参数约束，来实现正则化。
- l0 norm，$||w||_{0} = \\\#(i | w_i \ne 0 )$，用于稀疏编码(Sparse Encoding)等，有整数相关约束的场景；由于是NP难问题，往往通过Orthogonal Matching Pursuit(OMP)等贪婪算法求解

- l1 norm，$||w||_{1} = \sum_i |w_i|$，也称为Lasso

- l2 norm，$||w||_{2} = \sqrt{\sum_i w_i^2}$，也称为Ridge

- elastic net，$\lambda_1 \cdot ||w||_1 + \lambda_2 \cdot  ||w||_2$，以前用于工业界中上亿维度特征选择

5.核化(kernel)，任何点积形式的函数，在满足Mercer's Theorem条件下，可以进行核化，映射到Hilbert空间。那么我们可以想象"核化"的应用场景，包括相似性度量、协方差矩阵计算、模型计算(权重*特征)等，很多地方均可以进行核化。在学术圈也存在用核方法，对机器学习算法建模的研究领域(Adaptive Kernel Filtering)。<font color=red>狭义的核化，指将特征由低维空间映射到高维空间，解决低维空间不可分的问题。</font>
- 径向基核函数(高斯核，rbf)，原则上可以映射到无穷维，所以是通常选择。$K(x_{1},x_{2})=exp(-\gamma ||x_{1}-x_{2}||_2^2)$

- 线性核函数(linear)，映射后还在原空间。$K(x_{1},x_{2})=x_{1}^T x_{2} + C$

- 多项式核函数(polynomial)，有点像做特征组合。$K(x_{1},x_{2})=(\alpha x_{1}^T x_{2} + C)^d$

- Sigmoid核函数，$K(x_{1},x_{2})=tanh(\gamma x_{1}^{T} x_{2} + C)$

6.集成学习(Ensemble Learning)，通过将多个弱分类器结合一起使用，实现强分类器的效果，提升模型性能。按弱分类器间的依赖关系，可以分为bagging、boosting两种。<font color=red>近期在Kaggle、KDD Cup、天池等数据挖掘、机器学习竞赛中，以GBDT为首的集成学习算法，属于翘楚。</font>
- Bagging，指重复m轮，将训练集有放回的随机选取M个样本，进行训练得到m个弱分类器；<font color=red>弱分类器间是等权没有依赖关系，是并联关系</font>；最终预测结果，通过m个分类器进行投票所得。比如随机森林算法。

- Boosting，对同一份数据，训练m个弱分类器；<font color=red>相互之间是级联(Cascade)关系</font>，上一个弱分类器模型预测错的样本，会被放大权重，以保证收敛。比如AdaBoost、GBDT算法。

```python
# 接下来我们看下kernel、Boosting的能力，reference:http://scikit-learn.org/stable/auto_examples/classification/plot_classifier_comparison.html
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import make_moons, make_circles, make_classification
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import AdaBoostClassifier
# step size in the mesh
h = .02
# 1) 三种分类器，线性SVM，RBF核的SVM，以决策树为弱分类器的AdaBoosting
names = ["Linear SVM", "RBF SVM", "AdaBoost"]
classifiers = [SVC(kernel="linear", C=0.025),
    SVC(kernel="rbf", gamma=2, C=1),
    AdaBoostClassifier(base_estimator=None, n_estimators=50, learning_rate=1.0, algorithm='SAMME.R', random_state=0)]
# 2) 三种数据集，月亮型，圆形，线性可分
X, y = make_classification(n_features=2, n_redundant=0, n_informative=2, random_state=1, n_clusters_per_class=1)
rng = np.random.RandomState(2)
X += 2 * rng.uniform(size=X.shape)
linearly_separable = (X, y)
datasets = [make_moons(noise=0.3, random_state=0),
            make_circles(noise=0.2, factor=0.5, random_state=1),
            linearly_separable]
# 3) 画图展示
figure = plt.figure(figsize=(12, 6))
i = 1
# iterate over datasets
for ds_cnt, ds in enumerate(datasets):
    # preprocess dataset, split into training and test part
    X, y = ds
    X = StandardScaler().fit_transform(X)
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.4, random_state=42)
    # 4) 数据展示
    x_min, x_max = X[:, 0].min() - .5, X[:, 0].max() + .5
    y_min, y_max = X[:, 1].min() - .5, X[:, 1].max() + .5
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
    # just plot the dataset first
    cm = plt.cm.RdBu
    cm_bright = ListedColormap(['#FF0000', '#0000FF'])
    ax = plt.subplot(len(datasets), len(classifiers) + 1, i)
    if ds_cnt == 0:
        ax.set_title("Input data")
    # Plot the training points
    ax.scatter(X_train[:, 0], X_train[:, 1], c=y_train, cmap=cm_bright, edgecolors='k')
    # and testing points
    ax.scatter(X_test[:, 0], X_test[:, 1], c=y_test, cmap=cm_bright, alpha=0.5, edgecolors='k')
    ax.set_xlim(xx.min(), xx.max())
    ax.set_ylim(yy.min(), yy.max())
    ax.set_xticks(())
    ax.set_yticks(())
    i += 1
    # 5) 画分类器的分界平面，训练模型后，用模型预测每个meshgrid上的点，给出预测值
    # iterate over classifiers
    for name, clf in zip(names, classifiers):
        ax = plt.subplot(len(datasets), len(classifiers) + 1, i)
        clf.fit(X_train, y_train)
        score = clf.score(X_test, y_test) # score为精度
        # Plot the decision boundary. For that, we will assign a color to each point in the mesh [x_min, x_max]x[y_min, y_max].
        if hasattr(clf, "decision_function"):
            Z = clf.decision_function(np.c_[xx.ravel(), yy.ravel()])
        else:
            Z = clf.predict_proba(np.c_[xx.ravel(), yy.ravel()])[:, 1]
        # Put the result into a color plot
        Z = Z.reshape(xx.shape)
        ax.contourf(xx, yy, Z, cmap=cm, alpha=.8)
        # Plot also the training points，训练集颜色是纯色
        ax.scatter(X_train[:, 0], X_train[:, 1], c=y_train, cmap=cm_bright, edgecolors='k')
        # and testing points，测试集颜色是透明色
        ax.scatter(X_test[:, 0], X_test[:, 1], c=y_test, cmap=cm_bright, edgecolors='k', alpha=0.5)
        ax.set_xlim(xx.min(), xx.max())
        ax.set_ylim(yy.min(), yy.max())
        ax.set_xticks(())
        ax.set_yticks(())
        if ds_cnt == 0:
            ax.set_title(name)
        ax.text(xx.max() - .3, yy.min() + .3, ('%.2f' % score).lstrip('0'), size=15, horizontalalignment='right')
        i += 1

plt.tight_layout()
plt.show()
```
