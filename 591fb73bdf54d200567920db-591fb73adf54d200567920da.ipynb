# 当我们在买沪深300指数基金的时候，我们在买什么？

```python
import pandas as pd
import matplotlib.pyplot as plt
from CAL.PyCAL import *
font.set_size(22)
small_font = font.copy()
small_font.set_size(14)
```

```python
elements = DataAPI.IdxConsGet(secID=u"",ticker=u"000300",intoDate=u"",isNew=u"",field=u"",pandas="1")
weights = DataAPI.IdxCloseWeightGet(secID=u"",ticker=u"000300",beginDate=u"",endDate=u"",field=u"",pandas="1")
```

```python
elements.head(10)
```

```python
weights.head(10)
```

```python
elements = elements[['secID','consID','intoDate','consShortName']]
weights = weights[['effDate','consID','weight']]
```

```python
idx = pd.merge(elements,weights,on='consID',how='left').sort_values('effDate').groupby('consShortName').last().sort_values('consID')
idx = idx.reset_index()
```

```python
idx.head(10)
```

```python
ind_category = DataAPI.EquIndustryGet(industryVersionCD=u"010303",industry=u"",secID=u"",ticker=u"",intoDate=u"",field=u"",pandas="1")
ind_category = ind_category[ind_category['isNew'] == 1]
del ind_category['ticker']
del ind_category['exchangeCD']
del ind_category['secShortName']
del ind_category['secFullName']
del ind_category['partyID']
del ind_category['industryVersionCD']
del ind_category['industry']
del ind_category['outDate']
del ind_category['isNew']

ind_category.head(10)
```

```python
df = pd.merge(idx,ind_category,left_on = 'consID',right_on='secID')
```

```python
res = df.groupby('industryName1')['weight'].sum()
res.index.str.decode('utf8')
xticklabels = res.index
xticklabels = [i.decode('utf8') for i in xticklabels.values]
ax = res.plot(kind='bar',figsize=(18,8))
ax.set_xticklabels(xticklabels,fontproperties=small_font,rotation='vertical')
ax.set_xlabel(u'申万一级行业分类',fontproperties=small_font)
plt.show()
```

```python
res = df.groupby('industryName1')['weight'].sum().sort_values(0)

xticklabels = res.index.str.decode('utf8')
ax = res.plot(kind='bar',figsize=(18,8))
ax.set_xticklabels(xticklabels,fontproperties=small_font,rotation='vertical')
ax.set_xlabel(u'申万一级行业分类',fontproperties=small_font)
plt.show()
```
