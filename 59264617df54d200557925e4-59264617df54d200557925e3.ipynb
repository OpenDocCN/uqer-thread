	Series数据类型的创建

Series类型可以由如下类型创建：
• Python列表
• 标量值
• Python字典

• 其他函数

```python
#从标题创建
import pandas as pd
s=pd.Series(25,index=['a','b','c'])#注意不能省略index
s
```

```python
#从字典创建
import pandas as pd
d=pd.Series({'a':9,'b':8,'c':7})
print(d)
e=pd.Series({'a':9,'b':8,'c':7},index=['c','a','b','d']) #index从字典中进行选择操作
e
```

	Series类型的基本操作
Series类型包括index和values两部分
Series类型的操作类似ndarray类型
Series类型的操作类似Python字典类型

```python
#基本操作：获取索引与数据
import pandas as pd
b=pd.Series([9,8,7,6],['a','b','c','d'])
print(b)
print(b.index)
print(b.values)
b.index
```

```python
#基本操作：索引与切片

#可以通过自动序号索引或自定义索引。这两种索引并存，但不能混用。
import pandas as pd
b=pd.Series([9,8,7,6],['a','b','c','d'])
print(b)

print('\n')
print(b['b']) #使用自定义索引

print('\n')
print(b[1]) #使用自动索引，从0开始

#切片，连带索引一起切
x1=b[:3]
print('\nx1=')
print(x1)

x2=b[b>b.median()]
print('\nx2=')
print(x2)

x3=np.exp(b)
x2=b[b>b.median()]
print('\nx3=')
print(x3)
```

```python
#基本操作：in操作与 .get()方法
import pandas as pd
b=pd.Series([9,8,7,6],['a','b','c','d'])
print(b)

x1= 'c' in b
print('\nx1=')
print(x1)

x2= 0 in b  #不判断自动索引
print('\nx2=')
print(x2)

b.get('f',100) 返回键值，键值不存在，直接返回参数，没有参数返回空
```

```python
#基本操作：对齐操作
#Series类型在运算中会自动对齐不同索引的数据
import pandas as pd
a=pd.Series([1,2,3],['c','d','e'])
b=pd.Series([9,8,7,6],['a','b','c','d'])
a+b
```

	DataFrame类型
由共用相同索引的一组列组成
DataFrame是一个表格型的数据类型，每列值类型可以不同
DataFrame既有行索引、也有列索引
DataFrame常用于表达二维数据，但可以表达多维数据

DataFrame是二维带“标签”数组


	DataFrame类型的创建
• 二维ndarray对象
• 由一维ndarray、列表、字典、元组或Series构成的字典
• Series类型
• 其他的DataFrame类型

```python
#从二维ndarray对象创建
import pandas as pd
import numpy as np
d=pd.DataFrame(np.arange(10).reshape(2,5))   #注意行、列将生成自动索引
d
```

```python
#从列表类型的字典创建
import pandas as pd
d1={'one':[1,2,3,4],'two':[9,8,7,6]}
d=pd.DataFrame(d1,index=['a','b','c','d'])
d
```

	DataFrame类型的操作：重新索引

.reindex(index=None, columns=None, …)的参数
参数                  说明
index, columns       新的行列自定义索引
fill_value           重新索引中，用于填充缺失位置的值
method               填充方法, ffill当前值向前填充，bfill向后填充
limit                最大填充量
copy                 默认True，生成新的对象，False时，新旧相等不复制

```python
import pandas as pd
d1 = {'城市':['北京','上海','广州','深圳','沈阳'],
      '环比':[101.5, 101.2, 101.3, 102,  100.1],
      '同比':[120.7, 127.3, 119.4, 140.9,101.4],
      '定基':[121.4, 127.8, 120.0, 145.5,101.6]}
d=pd.DataFrame(d1,index=['c1','c2','c3','c4','c5'])
print d   #第一次使用到这里，显示数据有框线
print '\n'
x1=d.reindex(index=['c5','c4','c3','c2','c1'])  #同样的，此处如果是对列的索引重新排序，列将重新索引。
print x1  #第二次使用print，就没有框线了
x1        #不使用print，有框线
```

	索引类型的常用方法
方法 					  说明
.append(idx)         连接另一个Index对象，产生新的Index对象
.diff(idx)           计算差集，产生新的Index对象
.intersection(idx)   计算交集
.union(idx)          计算并集
.delete(loc)         删除loc位置处的元素
.insert(loc,e)       在loc位置增加一个元素e

	pandas数据类型的运算
算术运算法则：
算术运算根据行列索引，补齐后运算，运算默认产生浮点数
补齐时缺项填充NaN (空值)
二维和一维、一维和零维间为广播运算
采用+ ‐ * /符号进行的二元运算产生新的对象

```python
#算术运算
import pandas as pd
import numpy as np
a = pd.DataFrame(np.arange(12).reshape(3,4))
b = pd.DataFrame(np.arange(20).reshape(4,5))
x1=a+b
x2=a*b
print a
print '\n'
print b 
print '\n'
print x1 
print '\n'
print x2 
```
