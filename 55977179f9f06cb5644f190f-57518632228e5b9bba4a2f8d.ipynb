# 历史相似性

```python
import datetime as dt
import numpy as np
import pandas as pd
import seaborn as sns
sns.set_style('white')
from matplotlib import pylab
from CAL.PyCAL import *
font.set_size(20)

index = '000300'
data = DataAPI.MktIdxdGet(ticker = index, beginDate='20000101', endDate='20150702')
data.index = data.tradeDate.apply(lambda x: dt.datetime.strptime(x, '%Y-%m-%d'))

data2 = DataAPI.MktIdxdGet(ticker = index, beginDate='20140702', endDate='20150702')
data2.index = data2.tradeDate.apply(lambda x: dt.datetime.strptime(x, '%Y-%m-%d'))

data['corr'] = pd.rolling_apply(data['closeIndex'], len(data2), lambda x: np.corrcoef(x, data2['closeIndex'].values)[0, 1])

#data['corr'] = pd.rolling_corr(data['closeIndex'], data2['closeIndex'], window = 240)

data['corr'].plot(figsize=(16,8), grid = False)
sns.despine()
```
