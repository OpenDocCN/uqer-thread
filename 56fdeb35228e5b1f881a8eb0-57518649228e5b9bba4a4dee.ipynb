```python
#!usr/bin/env python
#-*- coding: utf-8 -*-
#Python机器学习库scikit-learn实践 - zouxy09的专栏 - 博客频道 - CSDN.NET http://blog.csdn.net/zouxy09/article/details/48903179
    
from sklearn import metrics
import numpy as np
import time


# Multinomial Naive Bayes Classifier 朴素贝叶斯分类器 http://blog.csdn.net/sulliy/article/details/6629201
def naive_bayes_classifier(train_x, train_y):
    from sklearn.naive_bayes import MultinomialNB
    model = MultinomialNB(alpha=0.01)
    model.fit(train_x, train_y)
    return model


# KNN Classifier #K邻近分类器 #http://scikit-learn.org/stable/modules/neighbors.html 
def knn_classifier(train_x, train_y):
    from sklearn.neighbors import KNeighborsClassifier
    model = KNeighborsClassifier(algorithm='auto')
    model.fit(train_x, train_y)
    return model


# Logistic Regression Classifier #logistic回归分类器
def logistic_regression_classifier(train_x, train_y):
    from sklearn.linear_model import LogisticRegression
    model = LogisticRegression(penalty='l2')
    model.fit(train_x, train_y)
    return model


# Random Forest Classifier #随机森林树 http://www.cnblogs.com/wentingtu/archive/2011/12/22/2297405.html
def random_forest_classifier(train_x, train_y):
    from sklearn.ensemble import RandomForestClassifier
    model = RandomForestClassifier(n_estimators=8)
    model.fit(train_x, train_y)
    return model


# Decision Tree Classifier #决策树
def decision_tree_classifier(train_x, train_y):
    from sklearn import tree
    model = tree.DecisionTreeClassifier()
    model.fit(train_x, train_y)
    return model


# GBDT(Gradient Boosting Decision Tree) Classifier #什么鬼
def gradient_boosting_classifier(train_x, train_y):
    from sklearn.ensemble import GradientBoostingClassifier
    model = GradientBoostingClassifier(n_estimators=200)
    model.fit(train_x, train_y)
    return model


# SVM Classifier #支持向量机 什么鬼
def svm_classifier(train_x, train_y):
    from sklearn.svm import SVC
    model = SVC(kernel='rbf', probability=True)
    model.fit(train_x, train_y)
    return model

# SVM Classifier using cross validation#交叉验证
def svm_cross_validation(train_x, train_y):
    from sklearn.grid_search import GridSearchCV
    from sklearn.svm import SVC
    model = SVC(kernel='rbf', probability=True)
    param_grid = {'C': [1e-3, 1e-2, 1e-1, 1, 10, 100, 1000], 'gamma': [0.001, 0.0001]}
    grid_search = GridSearchCV(model, param_grid, n_jobs = 1, verbose=1)
    grid_search.fit(train_x, train_y)
    best_parameters = grid_search.best_estimator_.get_params()
    for para, val in best_parameters.items():
        print para, val
    model = SVC(kernel='rbf', C=best_parameters['C'], gamma=best_parameters['gamma'], probability=True)
    model.fit(train_x, train_y)
    return model


def MLearning(train_x, train_y, test_x, test_y,predict_x,classifier):#return predict_y accurecy 
    #'NB', 'KNN', 'LR', 'RF', 'DT', 'SVM', 'GBDT'
    model = classifiers[classifier](train_x, train_y)
    predict = model.predict(test_x)
#     precision = metrics.precision_score(test_y, predict)
#     recall = metrics.recall_score(test_y, predict)  
    accuracy = metrics.accuracy_score(test_y, predict)
    predict_y = model.predict(predict_x)
    return predict_y,accuracy


def test():
    print 'reading training and testing data...'
    
    Index=DataAPI.MktIdxdGet(tradeDate=u"",indexID=u"",ticker=u"000001",\
                         beginDate=u"19900101",endDate=u"20160201",field=u"openIndex,turnoverVol,closeIndex",pandas="1")
    x=Index[['turnoverVol','openIndex']].values
    y=(Index['closeIndex']>Index['openIndex']).values
    divn=int(len(x)/2)
    train_x=x[divn:]
    train_y=y[divn:]
    test_x=x[0:divn]
    test_y=y[0:divn]
    
    predict_x=test_x[0:1]
    for classifier in test_classifiers:
        predict_y,accuracy=MLearning(train_x, train_y, test_x, test_y,predict_x,classifier)
        print classifier,predict_y,accuracy
    

##
test_classifiers = ['NB', 'KNN', 'LR', 'RF', 'DT', 'SVM', 'GBDT']
classifiers = {'NB':naive_bayes_classifier, 
              'KNN':knn_classifier,
               'LR':logistic_regression_classifier,
               'RF':random_forest_classifier,
               'DT':decision_tree_classifier,
              'SVM':svm_classifier,
            'SVMCV':svm_cross_validation,
             'GBDT':gradient_boosting_classifier
}    
    

```
