# 例题1-1：构建最优风险投资组合

马科维茨的均值-方差模型是现代投资组合理论的起点，模型中的有效边界更是金融工程中最重要的曲线之一。本文皆在展示如何利用Python绘制有效边界（Efficient Frontier），构建最优风险组合（Optimal Risky Portfolio）。求解ORP的数学形式本质是一个二次规划（Quadratic Programming）问题，在Python中可以利用cvxopt库解决这类问题，因而问题得以简化。

例1  请构建五种股票（'000002.XSHE'万科A、'000413.XSHE'东旭光电、'000063.XSHE'中兴通讯、'002007.XSHE'华兰生物、'000001.XSHE'平安银行）年（2016－2018），价格（已经包括了分红在内），假设你在2019年时有一笔资金准备投资这五种股票，并期望年收益率至少达到15％，那么你应当如何投资？当期望的年收益率变化时，投资组合和相应的风险如何变化？ 

```python
#------获取股票数据-------

import numpy as np
import scipy.optimize as sco
import pandas as pd
import matplotlib.pyplot as plt
start = '20160101'
end = '20180101'
universe = ['000413.XSHE','000063.XSHE','002007.XSHE']
data = pd.DataFrame()
for capitial in universe:
    ret = DataAPI.MktEqudGet(beginDate=start, endDate=end, secID=capitial,pandas='1',\
                            field=['tradeDate', 'secShortName', 'closePrice', 'preClosePrice'])
    ret['returns'] = ret.closePrice/ret.preClosePrice-1
    data = pd.concat([data,ret], axis=0)
data = pd.crosstab(data.tradeDate,data.secShortName,data.returns,aggfunc=max)
data = data.fillna(0)
data.head(5)
```

```python
#-------根据数据求有效边界，ORP，MVP-----------
exp_ret = data.mean() * 252
cov_mat = data.cov() * 252
risk_free = 0.04
sec_num = len(exp_ret)
def random_combination():
    sec_num = len(exp_ret)
    random_rets = []
    random_stds = []
    for i in range(4000):
        weights = np.random.random(sec_num)
        weights = weights / np.sum(weights)
        ret, std, _ = statistic(weights)
        random_rets.append(ret)
        random_stds.append(std)
    return(random_rets, random_stds)

def statistic(weights, exp_ret=exp_ret, cov_mat=cov_mat):
    sec_num = len(exp_ret)
    weights = np.asmatrix(weights).T
    exp_ret = np.asmatrix(exp_ret).T
    cov_mat = np.asmatrix(cov_mat)
    ret = float(weights.T * exp_ret)
    std = float(np.sqrt(weights.T * cov_mat * weights))
    sharpe = (ret - risk_free) / std
    return(ret, std, sharpe)

def efficiency_frontier(exp_ret=exp_ret, cov_mat=cov_mat):
    sec_num = len(exp_ret)
    min_ret = exp_ret.min()
    max_ret = exp_ret.max()
    rets = []
    stds = []
    for ret_level in np.linspace(min_ret, max_ret, 200):
        P = matrix(2 * cov_mat.values)
        q = matrix(np.zeros(sec_num))
        G = matrix(-np.eye(sec_num))
        h = matrix(np.zeros(sec_num))
        A = matrix(np.array([np.ones(sec_num),exp_ret.values]))
        b = matrix([1.0,ret_level])
        solvers.options['show_progress'] = False
        sol = solvers.qp(P,q,G,h,A,b)
        rets.append(ret_level)
        stds.append(np.sqrt(sol['primal objective']))
    return(rets, stds)
    
def min_sharpe(weights):
    return(-statistic(weights)[2])
from cvxopt import solvers, matrix
```

```python
cons = {'type':'eq', 'fun':lambda x:np.sum(x) - 1.0}
bnds = sec_num * ((0,1),)
opts = sco.minimize(min_sharpe, sec_num*[1.0/sec_num], method='SLSQP', bounds=bnds, constraints=cons)
orp_ret, orp_std, orp_sharpe = statistic(opts['x'])
rets, stds= efficiency_frontier()
random_rets, random_stds=random_combination()

fig = plt.figure(figsize=(14,8))
ax = fig.add_subplot(111)
plt.plot(0,risk_free,'k*',markersize=10)               #无风险资产
ax.plot(stds,rets)                                    #有效边界
ax.plot(orp_std,orp_ret,'r*',markersize=10)           #最优风险组合
plt.scatter(random_stds, random_rets, c=np.array(random_rets)/np.array(random_stds))  #随机组合
plt.colorbar(label='Sharpe Ratio')
plt.plot((0,orp_std),(risk_free,orp_ret),c='r')
ax.set_xlim(0,0.5)
```

```python
def get_historical_closes(ticker, start_date, end_date):
    # get the data for the tickers.  This will be a panel
    p = DataAPI.MktEqudGet(secID=ticker,beginDate=start_date,endDate=end_date,field=u"secID,tradeDate,closePrice",pandas="1") 
    p['tradeDate'] = p.apply(lambda row: (datetime.strptime(row["tradeDate"], "%Y-%m-%d")), axis = 1) #因为DataAPI所有日期都是以字符串形式存储的，因此此处需要转换成日期时间型,这里很重要！
    # rename the columns
    p.rename(columns={'tradeDate': 'Date', 
                      'closePrice': 'Close'}, inplace=True)
    # pivot each ticker to a column
    pivoted = p.pivot(index='Date', columns='secID')
    # and drop the one level on the columns
    pivoted.columns = pivoted.columns.droplevel(0)
    return pivoted
```

```python
# calculate annual returns
def calc_annual_returns(daily_returns):
    grouped = np.exp(daily_returns.groupby(
        lambda date: date.year).sum())-1
    return grouped
```

```python
annual_returns = calc_annual_returns(daily_returns)
annual_returns
```

```python
#Weighted variance co-variance calculation
```

```python
def calc_portfolio_var(returns, weights=None):
    if weights is None: 
        weights = np.ones(returns.columns.size) / \
        returns.columns.size
    sigma = np.cov(returns.T,ddof=0)
    var = (weights * sigma * weights.T).sum()
    return var
```

```python
# calculate our portfolio variance (equal weighted)
calc_portfolio_var(annual_returns)
```

```python
#The Sharpe Ratio
```

```python
def sharpe_ratio(returns, weights = None, risk_free_rate = 0.015):
    n = returns.columns.size
    if weights is None: weights = np.ones(n)/n
    # get the portfolio variance
    var = calc_portfolio_var(returns, weights)
    # and the means of the stocks in the portfolio
    means = returns.mean()
    # and return the sharpe ratio
    return (means.dot(weights) - risk_free_rate)/np.sqrt(var)
```

```python
# calculate equal weighted sharpe ratio
sharpe_ratio(annual_returns)
```

```python
#Optimization and minimization
```

```python
def negative_sharpe_ratio_n_minus_1_stock(weights, 
                                          returns, 
                                          risk_free_rate):
    """
    Given n-1 weights, return a negative sharpe ratio
    """
    weights2 = sp.append(weights, 1-np.sum(weights))
    return -sharpe_ratio(returns, weights2, risk_free_rate)
```

```python
def optimize_portfolio(returns, risk_free_rate):
    """ 
    Performs the optimization
    """
    # start with equal weights
    w0 = np.ones(returns.columns.size-1, 
                 dtype=float) * 1.0 / returns.columns.size
    # minimize the negative sharpe value
    w1 = scopt.fmin(negative_sharpe_ratio_n_minus_1_stock, 
                    w0, args=(returns, risk_free_rate))
    # build final set of weights
    final_w = sp.append(w1, 1 - np.sum(w1))
    # and calculate the final, optimized, sharpe ratio
    final_sharpe = sharpe_ratio(returns, final_w, risk_free_rate)
    return (final_w, final_sharpe)
```

```python
# optimize our portfolio
optimize_portfolio(annual_returns, 0.0003)
```

```python
#Visualizing the Efficient Frontier
```

```python
def objfun(W, R, target_ret):
    stock_mean = np.mean(R,axis=0)
    port_mean = np.dot(W,stock_mean) # portfolio mean
    cov=np.cov(R.T) # var-cov matrix
    port_var = np.dot(np.dot(W,cov),W.T) # portfolio variance
    penalty = 2000*abs(port_mean-target_ret)# penalty 4 deviation
    return np.sqrt(port_var) + penalty # objective function
```

```python
def calc_efficient_frontier(returns):
    result_means = []
    result_stds = []
    result_weights = []
    
    means = returns.mean()
    min_mean, max_mean = means.min(), means.max()
    
    nstocks = returns.columns.size
    
    for r in np.linspace(min_mean, max_mean, 100):
        weights = np.ones(nstocks)/nstocks
        bounds = [(0,1) for i in np.arange(nstocks)]
        constraints = ({'type': 'eq', 
                        'fun': lambda W: np.sum(W) - 1})
        results = scopt.minimize(objfun, weights, (returns, r), 
                                 method='SLSQP', 
                                 constraints = constraints,
                                 bounds = bounds)
        if not results.success: # handle error
            raise Exception(result.message)
        result_means.append(np.round(r,4)) # 4 decimal places
        std_=np.round(np.std(np.sum(returns*results.x,axis=1)),6)
        result_stds.append(std_)
        
        result_weights.append(np.round(results.x, 5))
    return {'Means': result_means, 
            'Stds': result_stds, 
            'Weights': result_weights}
```

```python
# calculate our frontier
frontier_data = calc_efficient_frontier(annual_returns)
```
