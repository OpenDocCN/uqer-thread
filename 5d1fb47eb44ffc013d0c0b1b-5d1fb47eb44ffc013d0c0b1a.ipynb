```python
import numpy as np
import pandas as pd
from scipy.stats.mstats import winsorize
import matplotlib.pyplot as plt

universe = set_universe('SH50') # 以上证50市盈率因子进行说明
# print(universe)
# MktStockFactorsOneDayGet ： 获取多只股票历史上某一天的因子数据
data = DataAPI.MktStockFactorsOneDayGet(tradeDate='20150916', secID=universe, field='ticker,PE',pandas='1').set_index('ticker')

"""
为缓解离群值对参数估计的影响，对属于连续型的解释变量在其分布的第1及99百分位上的观察值进行缩尾调整（Winsorize）处理。

    limits : {None, tuple of float}, optional
        Tuple of the percentages to cut on each side of the array, with respect
        to the number of unmasked data, as floats between 0. and 1.
        Noting n the number of unmasked data before trimming, the
        (n*limits[0])th smallest data and the (n*limits[1])th largest data are
        masked, and the total number of unmasked data after trimming
        is n*(1.-sum(limits)) The value of one limit can be set to None to
        indicate an open interval.
    inclusive : {(True, True) tuple}, optional
        Tuple indicating whether the number of data being masked on each side
        should be rounded (True) or truncated (False).
    inplace : {False, True}, optional
        Whether to winsorize in place (True) or to use a copy (False)
    axis : {None, int}, optional
        Axis along which to trim. If None, the whole array is trimmed, but its
        shape is maintained.
"""
data = data['PE'].to_dict()
# new_data = winsorize(data, limits=None, inclusive=(True, True), inplace=False,axis=None) # winsorize之后数据
new_data = winsorize(data, limits=(0.2,0.2)) # winsorize之后数据
print(new_data)

df = pd.DataFrame(data=0, index=map(lambda x: x[:6], universe),columns=['before winsorize','after winsorize'])
# df

df['before winsorize'][data.keys()] = data.values() # 对比两者数据进行展示
df['after winsorize'][new_data.tolist().keys()] = new_data.tolist().values() # ！！原有的new_data.values()出错！！
df.reset_index(inplace=True)
# df
print(df)

df.plot(figsize=(14,7))

# winsorize?
```
