```python
import matplotlib.pyplot as plt
import numpy as np 
import pandas as pd
import statsmodels.api as sm

stockdata=DataAPI.MktAHCompdGet(beginDate=u'20060101',endDate=u'20161231')
stock_ticker_list=stockdata['tickerA']
stock_ticker_list=np.unique(stock_ticker_list)
stockdata=DataAPI.MktAHCompdGet(ticker=stock_ticker_list)
stockdata=stockdata.dropna(axis=0,how='any')
print(len(stock_ticker_list))

coefficient=[]
const=[]
for i in range(0,len(stock_ticker_list)-1):
    ticker_temp=stock_ticker_list[i]
    stockdata_temp=stockdata[stockdata['tickerA']==ticker_temp]
    X=stockdata_temp['closePriceA']
    X.columns=['coefficient']
    Y=stockdata_temp['closePriceHA']
    reg_res=sm.OLS(Y,X).fit()
    coefficient.append(reg_res.params.closePriceA)
    print(reg_res.summary())
plt.hist(coefficient,bins=15,normed=True)
plt.show()
```
