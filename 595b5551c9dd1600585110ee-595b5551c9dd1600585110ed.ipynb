# 定期报告披露越早越好？——基于预约披露的实证分析

## 摘要
--------
最近市场半年报预约披露早的公司又火了一把，龙头天马科技3天2个板。市场上与学术上一直流传着一些声音，定期报告公布越早的公司通常业绩会更好；而定期报告公布的越晚的公司，可能由于业绩不好等原因，要做许多盈余管理，导致其制作报表与审计都需要更多的时间。
此处我们使用通联数据的预约披露数据，来看看预约披露时间的早晚对当期报表的利润是否有显著的影响。

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter 
from CAL.PyCAL import *    # CAL.PyCAL中包含font
cal = Calendar('China.SSE')
```

```python
pre_disclo_summary = DataAPI.EquReportPrediscloGet(secID=set_universe('A'),ticker=u"",beginDate=u"20080601",endDate=u"20170501",field=u"secID,preDate,endDate",pandas="1")
pre_disclo_summary = pre_disclo_summary.drop_duplicates(['endDate','secID'])
pre_disclo_summary = pre_disclo_summary.set_index(['endDate','secID'])
```

```python
pre_disclo_summary.sort_index(level=1).head()
```

如上我们得到了，每一只股票定期报告对应的预约时间，需要注意的是，16年以前都没有季报的预约时间，不知道是本身没有，还是数据库有所缺失。

```python
def get_accumu_q_growth():
    raw_data = DataAPI.FdmtISGet(ticker=u"",secID=set_universe('A'),reportType=['Q1','S1','CQ3','A'],endDate=u"",
                             beginDate='2007-01-01',
                             field=u"secID,publishDate,reportType,endDate,NIncome",pandas="1")
    def get_growth(x):
        if len(x) != 2:
            return x[['endDate','NIncome']].iloc[-1:-1].set_index(['endDate'])['NIncome'].pct_change()
        else:
            return x[['endDate','NIncome']].sort_values(['endDate']).set_index(['endDate'])['NIncome'].pct_change().dropna()
    
    analysis_summary = raw_data.groupby(['secID','publishDate','reportType']).apply(lambda x:get_growth(x))
    analysis_summary = analysis_summary.dropna()
    analysis_summary_reshape = analysis_summary.reset_index().rename(columns={'NIncome':'GrowthRate'})
    return analysis_summary_reshape
```

```python
accumu_q_growth = get_accumu_q_growth()
accumu_q_growth = accumu_q_growth.drop_duplicates(subset=['endDate','secID','GrowthRate'])
accumu_q_growth = accumu_q_growth.set_index(['endDate','secID'])
```

接着我们调用财报API，得到每一只股票每个定期报告对应的累计季度增长率。

```python
def get_growth(x):
    try:
        # print accumu_q_growth.ix[x[0],x[1]]['GrowthRate']
        return accumu_q_growth.ix[x[0],x[1]]['GrowthRate'].values[0]
    except:
        return np.NAN
```

```python
pre_disclo_summary['GrwothRate'] = pre_disclo_summary.index.map(lambda x:get_growth(x))
pre_disclo_summary['preDate_rank'] = pre_disclo_summary.groupby(level=0,group_keys=False).apply(lambda x:x['preDate'].rank())
```

```python
pre_disclo_summary.head()
```

如上我们得到定期报告各股票的排序与对应其的增长率，接下来，我们可以进行分组分析。

```python
def factor_grouper(summary_df,quantile_point,col = 'factor_value'):

    def quantile_function(df, quantile_point, col = col):
        #Get the quantile value
        #Select the data in the group that falls at or below the quantile value and return it
        def group_num(x):
            if quantile_point >= 3:

                if x <= df[col].quantile(1.0/quantile_point):
                    return 1
                for i in range(2,quantile_point):
                    if x <=df[col].quantile(1.0*i/quantile_point) and x > df[col].quantile(1.0*(i-1)/quantile_point):
                        return i
                if x > df[col].quantile(1.0*i/quantile_point):
                    return quantile_point
            elif quantile_point == 2:

                if x <= df[col].quantile(1.0/quantile_point):
                    return 1
                else:
                    return 2
            else:
                raise ValueError('quantile group num must larger than 2!')

        df['group'] = df[col].map(lambda x:group_num(x))
        return df
    return summary_df.groupby(level=0).apply(quantile_function,quantile_point)
```

```python
pre_disclo_grouper = factor_grouper(pre_disclo_summary,10,col = 'preDate_rank')
```

因为，增长率容易受到分母的影响，导致有部分数值可能很大，此处我们用各分组增长率的中位数代表整组的情况。如下所示。

```python
time_group_df = pre_disclo_grouper.groupby([pre_disclo_grouper.index.get_level_values(0),'group']).median()
```

```python
time_group_df.head(20)
```

将每个时间点的定期报告取平均值汇总

```python
time_group_df_summary = time_group_df.groupby(time_group_df.index.get_level_values(1)).mean()
```

```python
fig = plt.figure(figsize=(15,7))
ax = fig.add_subplot(111)
time_group_df_summary['GrwothRate'].plot(kind='bar',ax=ax,color='deepskyblue')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['right'].set_color('none')
ax.yaxis.set_ticks_position('left')
ax.spines['bottom'].set_position('zero')
ax.set_title(u"业绩增长率中位数平均", fontproperties=font, fontsize=18)
ax.grid()
```

## 总结
----------
从分组业绩增长率的情况来看，总体来说，这个说法是成立的，按预约披露时间的早晚分组，预约披露时间早的公司的业绩增长情况要显著好于预约披露时间晚的公司，并且分组的单调性非常好。
