# 6.MACD简单择时--基于优矿

# MACD

---
- MACD（Moving Average Convergence and Divergence)是Geral Appel于1970年提出的，利用收盘价的短期（常用为12日）指数移动平均线与长期（常用为26日）指数移动平均线之间的聚合与分离状况，对买进、卖出时机作出研判的技术指标。由快的指数移动平均线（EMA12）减去慢的指数移动平均线（EMA26）得到快线DIF，由DIF的9日加权移动均线得到慢线DEA，再用2×（快线DIF-慢线DEA）得到MACD柱。
- MACD为正，说明DIF在DEA之上；MACD为负，说明DIF在DEA之下。所以当MACD由负转正时，即快线DIF上穿DEA，是买入的信号，反之当MACD由正转负时则是卖出的信号。

## 策略逻辑

---
- 当DIF上穿DEA，即MACD由负转正时，为买入信号；同时DIF>0，信号更强，买入股票。
- 当DIF下穿DEA，即MACD由正转负时，卖出股票。
- 同时控制仓位，每只股票的买入成本不超过100000。

- 一般计算MACD的三个参数为12、26、9，这里参考《浙商证券-技术指标优化择时：10年30倍收益》，将参数改为22、44、19，确实提高了策略收益。

## 总结
- MACD为趋势型策略，在不做空的情况下，仅在单边上涨的市场中表现较好，在震荡型市场中表现很一般，而且还是在优化了参数和加入DIF>0的条件后的效果，这两者在本例中都是使回测结果更好的改动。
- 作为趋势型指标，在单独使用的情况下，MACD是远不如SMA、EMA等均线指标的。所以以后还是不要单独使用MACD来做投资了。。。
