# 请教各位如何取一条曲线中的两个最高点？——根据大师的“双均线策略”

```python
from matplotlib import pylab
import numpy as np
import pandas as pd
import DataAPI
import seaborn as sns
sns.set_style('white')

secID = '300115.XSHE'
start = '20150101'
end = '20150814'

security = DataAPI.MktEqudGet(secID, beginDate=start, endDate=end, field=['tradeDate', 'closePrice'])
security['tradeDate'] = pd.to_datetime(security['tradeDate'])
security = security.set_index('tradeDate')
security.info()
security['closePrice'].plot(grid=False, figsize=(12,8))
sns.despine()
```
