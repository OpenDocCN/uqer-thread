## 读数据，回测

```python
def todayDataRecovery(universe):
    all_multi_factor_data={}
    for i,stock in enumerate(universe):
        file_name = 'UQER' +stock + '.csv'
        try:
            all_multi_factor_data[stock]= pd.read_csv(file_name)
            all_multi_factor_data[stock]= all_multi_factor_data[stock].iloc[:,1:]
        except:
            print 'wrong with stock ',stock
    print 'finsih todayDataRecovery'
    return all_multi_factor_data

def todayData(universe,trading_day,tempdata):
    all_multi_factor_data={}
    for i,stock in enumerate(universe):
        try:
            temp=tempdata[stock]
            temp=temp[temp.tradeDate==trading_day]
            all_multi_factor_data[stock]=temp
        except :
            print 'wrong with stock is ',stock
            continue    
    print 'finish read',trading_day
    return all_multi_factor_data

def DataRecovery(universe):
    all_multi_factor_data={}
    for i,stock in enumerate(universe):
        file_name = 'wsUQER' +stock + '.csv'
        try:
            all_multi_factor_data[stock]= pd.read_csv(file_name)
            all_multi_factor_data[stock]= all_multi_factor_data[stock].iloc[:,1:]
        except:
            print 'wrong with stock ',stock
    return all_multi_factor_data

#取前复权数据
def StockAdjRegDataGet(stock, trading_days, multi_factor_data , shift = 11,used_factors=''): #股票代码，假定交易日，单只股票全部多因子数据，回归预测时间
    start = trading_days[0]
    end  = trading_days[-1]
    data  = multi_factor_data[(multi_factor_data.tradeDate >= start.strftime('%Y-%m-%d')) & (multi_factor_data.tradeDate <= end.strftime('%Y-%m-%d'))][:-shift]
    #ret = DataAPI.MktAdjEqudGet(secID = stock, beginDate = start.strftime('%Y%m%d'), endDate = end.strftime('%Y%m%d'), field = ['tradeDate', 'closePrice'])
    ret = DataAPI.MktEqudGet(secID = stock, beginDate = start.strftime('%Y%m%d'), endDate = end.strftime('%Y%m%d'), field = ['tradeDate', 'closePrice'])
    ret['fwdPrice'] = ret['closePrice'].shift(-shift)
    ret['return'] = ret['fwdPrice'] / ret['closePrice'] - 1.
    ret = ret[:-shift]
    data = data.merge(ret, how = 'inner', left_on = ['tradeDate'], right_on = ['tradeDate'])
    data = data.loc[:, ['return'] + used_factors]
    return data

def GetRegressionResult(multi_factor_data,used_factors): #单只股票的多因子数据，多因子列表
    tmp=multi_factor_data
    tmp = tmp.dropna()# 去掉空值
    for fac in used_factors:
        if tmp[fac].std() == 0:
            continue
    y=np.array(tmp['return'].tolist())#tolist()方法可把数组转换为列表
    x = []#设置空列表
    for fac in used_factors:
        x.append(tmp[fac].tolist())
    x = np.column_stack(tuple(x))#一维数组堆叠成一个二维数组。
    x = np.array( [ np.append(v,1) for v in x] ) 
    
    en = ElasticNet(fit_intercept=True, alpha=0)
    en.fit(x, y)
    res = en.coef_[:-1]
    
    w = dict(zip(tmp, res))
    return w

def GetReadyGo(universe,date,all_multi_factor_data,shift = 6,used_factors=''):# 股票池，日期（输入为一天的日期），全股票池股票多因子数据panel结构，使用股票代码dict，日期、多因子dataframe
    date=Date(date.year,date.month,date.day)#CAL.PyCAL导入函数，变换时间格式
    cal = Calendar('China.SSE')#上海证券交易所交易日期
    start = cal.advanceDate(date, '-90B', BizDayConvention.Following)#从今天开始往前推3年约780个交易日，然后希望那天是个工作日，如果不是的话顺延
    end  = cal.advanceDate(date, '-1B', BizDayConvention.Following)
    start = datetime(start.year(), start.month(), start.dayOfMonth())
    end  = datetime( end.year(), end.month(), end.dayOfMonth())
    trading_days = quartz.utils.tradingcalendar.get_trading_days(start, end) #生成一个日期（datetime格式）返回list
    
    datas, means, vols, weights = {}, {}, {}, {}
    for i,stock in enumerate(universe):
        try:
            datas[stock]=StockAdjRegDataGet(stock, trading_days, all_multi_factor_data[stock], shift = 6,used_factors=used_factors) #股票代码，假定交易日，单只股票全部多因子数据，回归预测时间
            means[stock]   = dict(datas[stock].mean())
            vols[stock]    = dict(datas[stock].std())
            weights[stock] = GetRegressionResult(datas[stock],used_factors)
        except Exception, e:
            pass
    return means, vols, weights 
```
