```python
import numpy as np
import pandas as pd
import os
import scipy.stats as st
import datetime
from CAL.PyCAL import *
universe = set_universe('A')
cal = Calendar('China.SSE')
```

```python
raw_data_path = './Factor2'
if not os.path.exists(raw_data_path):
    os.mkdir(raw_data_path)    
```

```python
#定义所需要的各种函数：


#定义时间格式转换函数
def timechange(x):
    y = datetime.datetime.strptime(x,"%Y-%m-%d")
    z = y.strftime('%Y%m%d')
    return z
#timechange('2018-12-09')


#获取整个交易期间的灭一个交易日历，月初，月末
def get_trade_date(begin_date,end_date):
    cal = Calendar('China.SSE')
    cal_date = DataAPI.TradeCalGet(exchangeCD=u"XSHG",beginDate=begin_date,endDate=end_date,\
                                   field=u"calendarDate,isOpen,isMonthEnd",pandas="1")
    
    trade_date = cal_date[cal_date['isOpen']==1]['calendarDate'].tolist()
    trade_date = [timechange(x) for x in trade_date]
    
    month_end = cal_date[cal_date['isMonthEnd']==1]['calendarDate'].tolist()
    month_end = [timechange(x) for x in month_end]
    
    month_begin = [cal.advanceDate(x,'1B').strftime('%Y%m%d') for x in month_end]
    
    return trade_date,month_end,month_begin

#定义去掉ST函数
def ST_remove(secID_source,st_time = None):
    st_time = st_time if st_time is not None else datetime.datetime.now().strftime('%Y%m%d')
    ST_secID = DataAPI.SecSTGet(beginDate=st_time,endDate=st_time,secID=secID_source,\
                                ticker=u"",field=u"secID",pandas="1")
    ST_secID_list = list(ST_secID['secID'])
    return[x for x in secID_source if x not in ST_secID_list]

#定义去掉新股
def new_remove(secID_source,new_time = None,days = 20):
    
    new_time = new_time if new_time is not None else datetime.datetime.now().strftime('%Y%m%d')
    ipo_info = DataAPI.SecIDGet(partyID=u"",ticker=secID_source,cnSpell=u"",assetClass=u"e",\
                                exchangeCD="",listStatusCD="",field=u"ticker,listDate",pandas="1")
    period = '-'+str(days)+'B'
    date_time_ipo = cal.advanceDate(new_time,period).strftime('%Y%m%d')
    new_ipo_info = ipo_info[ipo_info['listDate']>=date_time_ipo]['ticker'].tolist()
    
    return [x for x in secID_source if x not in new_ipo_info]

#定义将代码形式转化为内部编码
def ticker2secID(ticker_source):
    all_code=DataAPI.EquGet(equTypeCD=u"A",listStatusCD=u"L,S,DE,UN",\
                            exchangeCD="",ListSectorCD=u"",field=u"secID,ticker",pandas="1")
    all_code = dict(all_code.set_index(all_code['ticker'])['secID'])
    all_secID = []
    if isinstance(ticker_source,list):
        for i in ticker_source:
            if i in all_code.keys():
                all_secID.append(all_code[i])
            else:
                print(i+'不在存储当中')
    else:
        print('请输入LIST')
    return all_secID 

#获得每月月初的股票池
def get_stock_pool(date_month_begin,N=50):
    
    univ = DynamicUniverse('A')
    all_code = univ.preview(date_month_begin,skip_halted=False)
    
    all_code = ST_remove(all_code,st_time = date_month_begin)
    all_code = [x.split('.')[0] for x in all_code]
    all_code = new_remove(all_code,new_time =date_month_begin ,days = N)
    all_code = ticker2secID(all_code)
    
    all_code_dict = {'code':all_code}
    all_code_df = pd.DataFrame(all_code_dict)
    all_code_df['date'] = date_month_begin 
    
    all_code_df = all_code_df[['date','code']]
    return all_code_df     
```

```python
#获得所有交易日月初的股票代码
begin_date = '20071220'
end_date = '20180601'
trade_date,month_end,month_begin = get_trade_date(begin_date,end_date)

all_stock = pd.DataFrame()
for i in month_begin:
    
    all_code_df = get_stock_pool(i,N=180)
    
    all_stock = pd.concat([all_stock,all_code_df],axis =0)
    
all_stock
```

```python
all_stock.to_csv('Factor2/stock_pool.csv',index = True)
all_stock2 = pd.read_csv('Factor2/stock_pool.csv',index_col= 0)
```

```python
# 获取整个交易期间所有的股票信息
all_code = sorted(list(set(list(all_stock['code']))))
mkt_info_frame = DataAPI.MktEqudGet(secID=all_code,beginDate=u"20071220",\
                                    endDate=u"20180601",isOpen="1",field=u"secID,tradeDate,chgPct,turnoverRate")
mkt_info_frame['tradeDate']=map(timechange,mkt_info_frame['tradeDate'])
mkt_info_frame.columns = ['code','tradeDate','ret','to']
mkt_info_frame = mkt_info_frame.dropna()
mkt_info_frame  #按照code 和日期共同的顺序进行排序
```

```python
mkt_info_frame.to_csv('Factor2/mkt_info_frame.csv',index = True)
mkt_info_frame2 = pd.read_csv('Factor2/mkt_info_frame.csv',index_col= 0)
```

```python
all_signal = pd.DataFrame()

for i in range(1,len(month_end)):
    begin = month_begin[i-1]
    end = month_end[i]
    start = month_begin[i]
    
    if i < len(month_end)-1:
        next_end = month_end[i+1]
    else:
        next_end = '20180629'
        
    # 获取每一个月末的下一个月的收益率
    code = sorted(all_stock[all_stock['date']==start]['code'].tolist())  # 以下一个月的第一天代码为标准
    
    info1 = DataAPI.MktEqudAdjGet(secID=code,tradeDate=end,isOpen=1,field=u"secID,closePrice",pandas="1")
    info1.columns = ['code','closeprice1']
    
    info2 = DataAPI.MktEqudAdjGet(secID=code,tradeDate=next_end,isOpen=1,field=u"secID,closePrice",pandas="1")
    info2.columns = ['code','closeprice2']
    
    info = pd.merge(info1,info2,on = 'code')
    
    info['month_ret'] = info['closeprice2']/info['closeprice1']-1
    info['date'] = end
    info = info[['date','code','month_ret']]
    
    #获取三个因子
    period_frame = mkt_info_frame[(mkt_info_frame['tradeDate']>=begin)&(mkt_info_frame['tradeDate']<=end)]
    
    period_frame =period_frame[period_frame['code'].isin(code)]
    
    period_frame['next_ret'] = period_frame.groupby('code')['ret'].shift(-1,axis = 0)
    period_frame =period_frame.dropna()
    
    #fr
    period_frame['abs_next_ret'] = abs(period_frame['next_ret'])
    
    n_days = len([x for x in trade_date if x>=begin and x<=end])/2    
    signal_fr = period_frame.groupby('code').apply(lambda df:df['to'].corr(df['abs_next_ret'],method = 'pearson',min_periods =n_days))
    
    #fru
    period_frame_positive = period_frame[period_frame['next_ret']>0]
    signal_fru =  period_frame_positive.groupby('code').apply(lambda df:df['to'].corr(df['next_ret'],method = 'pearson',min_periods =5))
    
    #fru
    period_frame_negative = period_frame[period_frame['next_ret']<=0]
    signal_frd =  period_frame_negative.groupby('code').apply(lambda df:df['to'].corr(df['next_ret'],method = 'pearson',min_periods =5))
     
    #三个因子结合到一起
    signal = pd.concat([signal_fr,signal_fru],axis = 1,join = 'outer')
    signal = pd.concat([signal,signal_frd],axis = 1,join = 'outer')
    signal.columns = ['fr','fru','frd']
    signal['date'] = end
    signal['code']= signal.index
    signal.reset_index(drop = True,inplace = True)
    #signal = signal.reset_index(drop = True,inplace = True)  reset_index()方法前面不用等号
    signal = signal[['date','code','fr','fru','frd']]
    
    signal = pd.merge(info,signal,how = 'right')
    all_signal = all_signal.append(signal)
    
    
    
all_signal.sort_values(['date','code'],inplace = True)
all_signal
```

```python
all_signal.to_csv('Factor2/all_signal.csv',index = True)
all_signal2 = pd.read_csv('Factor2/all_signal.csv',index_col= 0)
```

```python
#计算各个因子的IC值
def cal_ic(data,factor_name):
    factor_ic = all_signal.groupby('date').apply(lambda df:df['month_ret'].corr(df[factor_name],method = 'spearman'))
    factor_ic = pd.DataFrame(factor_ic)
    factor_ic['date'] = factor_ic.index
    factor_ic.reset_index(drop=True,inplace=True)
    factor_ic.columns = [factor_name,'date']
    factor_ic.sort_values('date',inplace=True)
    factor_ic = factor_ic[['date',factor_name]]
    return factor_ic

factor_ic_fr = cal_ic(all_signal,'fr')
factor_ic_fru = cal_ic(all_signal,'fru')
factor_ic_frd = cal_ic(all_signal,'frd')

raw_IC = pd.merge(factor_ic_fr,factor_ic_fru,on = 'date')
raw_IC = pd.merge(raw_IC,factor_ic_frd,on = 'date')
raw_IC
```
