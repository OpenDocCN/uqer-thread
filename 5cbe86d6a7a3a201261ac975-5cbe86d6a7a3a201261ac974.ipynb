**4、df_marketvalue from DataAPI.MktEqudGet**
factors_marketvalue_uqer = ['marketValue']

```python
begindate = '20070101'
enddate = '20190331'

factors_marketvalue_uqer = ['ticker','tradeDate','marketValue']
```

```python
df_marketvalue = DataAPI.MktEqudGet(ticker=train_ticker_list,beginDate=begindate,endDate=enddate,field=factors_marketvalue_uqer,pandas="1")
```

```python
df_marketvalue = df_marketvalue.sort_values(['ticker','tradeDate'],ascending=[True,True])
df_marketvalue = df_marketvalue.drop_duplicates(['ticker','tradeDate'],keep='first')
print (len(set(df_marketvalue['ticker'])))
print (df_marketvalue.head(10).to_html())
# print (df_marketvalue.loc[df_marketvalue['ticker']=='000002'][:10].to_html())
```

```python
# 数据比较大，保存起来，以免每次运行耗时
df_marketvalue.to_hdf('20070101_20190331_marketvalue.h5',key='marketvalue', mode='w')
```

**5、industries_frame from DataAPI.EquIndustryGet**
columns_industry_uqer = ['ticker','endDate','isNew','industryName1']

```python
# 获取行业信息
industries_frame = DataAPI.EquIndustryGet(industryVersionCD=u"010303", field=[u'ticker', 'secShortName', 'industry', 'industryName1','isNew'],intoDate=enddate, pandas="1")
industries_frame = industries_frame[industries_frame.isNew == 1]
print (industries_frame.shape)
print (industries_frame.head().to_html())
```

```python
# 市值与行业信息merge
df_marketvalue_industryname = pd.merge(df_marketvalue,industries_frame[['ticker','secShortName','industryName1']],on = 'ticker',how='left')
df_marketvalue_industryname['tradeDate'] = pd.to_datetime(df_marketvalue_industryname['tradeDate'])    
df_marketvalue_industryname['tradeDate_year&month'] = df_marketvalue_industryname['tradeDate'].dt.year.astype(str) + df_marketvalue_industryname['tradeDate'].dt.month.astype(str)
df_marketvalue_industryname = df_marketvalue_industryname.sort_values(['ticker','tradeDate'],ascending=[True,True]).drop_duplicates(['ticker','tradeDate_year&month'],keep='last')
print (df_marketvalue_industryname.shape)
print (df_marketvalue_industryname.head().to_html())
```

```python
# 三大报表因子与市值行业merge
merge_temp1 = df_cash.merge(df_income,on=['ticker','endDate','reportType','fiscalPeriod'],how='outer')
merge_temp2 = merge_temp1.merge(df_balance,on=['ticker','endDate','reportType','fiscalPeriod'],how='outer')
merge_temp2['endDate'] = pd.to_datetime(merge_temp2['endDate'])   
merge_temp2['endDate_year&month'] = merge_temp2['endDate'].dt.year.astype(str)+merge_temp2['endDate'].dt.month.astype(str) 

factor_merge = merge_temp2.merge(df_marketvalue_industryname, left_on=['ticker', 'endDate_year&month'], right_on=['ticker', 'tradeDate_year&month'], how='outer') 
print (factor_merge.shape)
factor_merge = factor_merge[factor_merge['fiscalPeriod'].notnull()]  #保留三大报表有的信息
factor_merge = factor_merge[factor_merge['industryName1'].notnull()]
print (factor_merge.shape)
print (factor_merge.head().to_html())
```

```python
# 三大报表因子与市值行业merge后保存起来，后续使用可以直接读取。周期为20070101-20190331
factor_merge.to_hdf('20070101_20190331_factor_merge.h5',key='factor_merge', mode='w')
```
