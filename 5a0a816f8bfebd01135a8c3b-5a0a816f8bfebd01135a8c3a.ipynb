# PEG选股策略研究

## 前言
---
* PEG指标即市盈率相对盈利增长比率，是由上市公司的市盈率除以盈利增长速度得到的数值。该指标既可以通过市盈率考察公司目前的财务状况，又通过盈利增长速度考察未来一段时期内公司的增长预期，因此是一个比较完美的选股参考指标。
    
* 该指标最先由英国投资大师史莱特提出，后来由美国投资大师彼得·林奇发扬光大。史莱特被称为PEG选股法的创始者，同时名列18位投资大师排行榜中。PEG指标最先在英国证券市场上使用，但是英国证券市场的影响力较小，所以该指标提出后，并未在世界引起较大的反响。1992年，史莱特通过自己的著作《祖鲁原则》，将市盈率相对盈利增长比率(PEG)这一投资和选股方法推广到了美国。此后，在美国著名投资大师彼得·林奇的努力下，终于使这一投资理念深入人心。后来，更多的人将PEG方法的功绩归功于彼得·林奇。
    
* 用PEG指标选股的好处就是将市盈率和公司业绩成长性对比起来看，其中的关键是要对公司的业绩作出准确的预期。投资者普遍习惯于使用市盈率来评估股票的价值，但是，当遇到一些极端情况时，市盈率的可操作性就有局限，比如市场上有许多远高于股市平均市盈率水平，甚至高达上百倍市盈率的股票，此时就无法用市盈率来评估这类股票的价值。但如果将市盈率和公司业绩成长性相对比，那些超高市盈率的股票看上去就有合理性了，投资者就不会觉得风险太大了，这就是PEG估值法。
    
* 本文主要参考了2011年1月9日广发证券的一篇题为《投资策略专题报告：PEG给力吗？——对年度策略建议的再审视》的研究报告，对近年来PEG的有效性进行了分析，并设计了PEG投资策略，分别对其短期和长期收益进行了对比。 

## 一、	市场对行业成长性的一致预期
---
* 首先，我们考察了市场对行业成长性的一致预期。从净利润增速的角度来看，市场预期2017年增长最快的行业是有色金属、采掘和钢铁；增速较慢的行业是银行、公用事业、房地产、建筑装饰和非银金融。
    
* 从复合增长率看，市场预期未来两年增长最快的行业是：电子、国防军工、采掘、有色金属和钢铁；未来两年净利润复合增长率较低的行业是建筑装饰、家用电器、公用事业和公用事业。大多数行业未来两年的业绩增速较2016年有较大幅度下滑，市场预期能在三年内持续保持较快增长的行业只有采掘、有色金属和钢铁。

```python
import pandas as pd
import numpy as np
import time
from CAL.PyCAL import *
import matplotlib.pyplot as plt
import matplotlib as mpl 

#读取2017年2月1日对于各行业在2016、2017和2018年的PE、同比增速、ROE、PB等一致预期
#该excel可以通过API得到：DataAPI.GG.SWconForeGGGet(ticker=u"申万行业代码xxx",industry=u"",beginDate=u"20170201",endDate=u"20170201",tmStamp=u"",field=u"",pandas="1")
raw_data = pd.read_excel(u'20170201_行业数据.xlsx')

#只考虑申万一级行业，删除其它行业分类数据
industries = [u'采掘', u'传媒', u'电气设备', u'电子', u'房地产', u'纺织服装', u'非银金融', u'钢铁', u'公用事业', u'国防军工', u'化工', u'机械设备', u'计算机', u'家用电器', u'建筑材料', u'建筑装饰', u'交通运输', u'农林牧渔', u'汽车', u'轻工制造', u'商业贸易', u'食品饮料', u'通信', u'休闲服务', u'医药生物', u'银行', u'有色金属', u'综合']
data = raw_data[raw_data.STOCK_NAME.isin(industries)]

data = data.drop_duplicates(['STOCK_NAME','RPT_DATE'])
data = data[['STOCK_CODE', 'STOCK_NAME','CON_DATE','RPT_DATE','C4','C5','C7','C12','CB','CPB']]
data.columns = ['Code', 'Industry','Date','Year','Earning','PE (2016E)','Growth','ROE','BV','PB (2016E)']

#基于同比增速和ROE将Long Dataframe转化为Short Dataframe
pivoted = data.pivot(index='Industry', columns='Year')
earning = pivoted['Growth']
earning.columns = ['Growth rate (2015)', 'Growth rate (2016E)','Growth rate (2017E)','Growth rate (2018E)']

roe = pivoted['ROE']
roe.columns = ['ROE (2015)', 'ROE (2016E)','ROE (2017E)','ROE (2018E)']

#计算未来两年复合增长率、PEG和PB/ROE
data = data.set_index('Industry')
data_processed = pd.merge(data[data['Year'] == 2016],earning,left_index=True, right_index=True)
data_processed = pd.merge(data_processed, roe,left_index=True, right_index=True)
data_processed['Compound growth rate (2017E & 2018E)'] = [np.sqrt((data_processed['Growth rate (2017E)'][i]+1)*(data_processed['Growth rate (2018E)'][i]+1))-1 for i in range(0,len(data_processed))] 
data_processed['PEG'] = data_processed['PE (2016E)']/data_processed['Compound growth rate (2017E & 2018E)']/100
data_processed['PB/ROE'] = data_processed['PB (2016E)']/data_processed['ROE']

data_show = data_processed[['PEG', 'Compound growth rate (2017E & 2018E)', 'PE (2016E)', 'Growth rate (2016E)', 'Growth rate (2017E)', 'Growth rate (2018E)']]
data_show.columns = ['PEG (PE/CGR)', 'CGR (2017E & 2018E)', 'PE (2016E)', 'GR (2016E)', 'GR (2017E)', 'GR (2018E)']
data_show
```

## 二、	行业成长性与估值的关系
---
* 接下来，我们探讨对行业成长性的预期是否会对该行业的估值水平（PE）产生影响。下面的3个图分别展示了用不同年份的预期增长率对PE进行回归得到的结果。其中，前两个图中分别使用 了2016年和2017年的预期同比增速（Growth rate, GR）作为解释变量，而第3个图则使用了未来两年（2017年和2018年）的预期年复合增长率（Compound growth rate, CGR）。各图中的直线分别为以相应解释变量对被解释变量PE进行回归后得到的直线。显然，图1中的直线拟合效果最差，图3中的直线拟合效果最好。因此，我们可以得到以下结论：
    
* 对行业成长性的预期，很大程度上决定了该行业的估值水平（PE）:
	* 2016年的业绩增速，对行业估值几乎没有影响。
	* 2017年的业绩增速很好地解释了不同行业的估值水平，因此估值很大程度上反映了市场对未来一年的增长预期。
	* 2017-2018年的复合增长率进一步提高了对估值的拟合优度。因此市场给某个行业估值时，会考虑未来两年的盈利增速。从拟合优度的角度看，考虑复合增长率之后，拟合优度仅略有提高，说明影响估值水平的最主要因素，仍然是未来一年的盈利增长。
    
* 根据以上分析，我们认为：（1）过去的业绩不会影响行业估值；（2）未来一年（t+1）的业绩增速是影响估值水平的最重要因素；（3）估值水平在一定程度上，也会考虑到未来第二年（t+2）的业绩增速，但其影响很小。
    
* 既然PE是行业成长性的反映，那么就可以根据估值和成长性的关系来确定某些行业的估值是否已经过高或过低，其衡量标准即PEG。

```python
#用各年份同比增速对PE作回归，并作出拟合直线
def Plot(x_name, y_name):
    model = pd.ols(y=data_processed[y_name], x=data_processed[x_name])
    plt.scatter(data_processed[x_name], data_processed[y_name], c='w',marker='o')
    ax = plt.axis()
    x = np.linspace(ax[0], ax[1]+0.01)
    plt.plot(x, model.beta[1]+model.beta[0]*x,'r',lw=2)
    plt.axis('tight')
    plt.xlabel(x_name) 
    plt.ylabel(y_name)
        
Plot('Growth rate (2016E)', 'PE (2016E)') 
```

## 三、	盈利能力与估值的关系
---
* 在估值中只考虑成长性已经足够了吗？盈利能力也会影响估值水平，因此还需要考虑基于PB-ROE的估值方法。由于PB等于PE乘以ROE，PB-ROE通常也被视为行业选择和比较的标准。一般而言，假定PE是外生的，市场愿意赋予盈利能力较强（从而ROE较高）的行业或公司较高的估值，因此PB与ROE应该是一种正相关的关系。相应的，高ROE、低PB的行业很可能被低估；低ROE、高PB的行业则可能被高估。
    
* 类似地，现在我们将研究对行业盈利能力的预期是否会对该行业的估值水平（PB）产生影响。下面的3个图分别展示了用不同年份的ROE预期对PB进行回归得到的结果。显然，各图的直线拟合效果均较差。因此，我们可以得到以下结论：ROE对行业估值的影响不明显，市场似乎仍不重视盈利能力对股指的重要影响。
    
* 根据一致预期，我们发现不同行业之间的PB与ROE之间存在正相关关系，但十分微弱，说明PB-ROE并非市场主流的估值方法，盈利能力对估值的重要影响似乎还未被认识到。可能的原因主要有以下几点：第一，行业研究员在对个股进行定价时，主要根据其成长性并给出相应的合理PE；第二，投资者在构建投资组合时，甚少关注PB和ROE之类的指标；第三，在同一行业的不同公司中，ROE对PB有显著影响，但由于行业之间的异质性，将不同行业放在一起之后其规律反而不明显。

## 四、	PEG的有效性
---
* 上述分析表明，我们可以使用PEG来衡量行业的估值水平。PEG显著高于合理水平的行业存在高估风险，反之PEG显著低于合理水平的行业存在低估的机会。
* 为此，我们确定了两个指标：（1）使用未来两年净利润的复合增长率作为PEG中的“G”；（2）以各行业的PEG均值作为PEG的合理中枢，以中枢的上下一倍标准差作为合理波动范围。
* 根据这一PEG中枢，一致预期数据表明：
	* 被明显低估的行业是房地产和汽车；
	* 被明显高估的行业是综合、休闲服务和国防军工；
	* PEG合理但已经较低的行业是钢铁、化工和建筑装饰；
	* 医药生物、传媒和商业贸易等行业的PEG虽然仍处于合理区间，但已经较高。
* 既然PEG在很大程度上能够有效定价，那么根据PEG的行业选择就变得十分简单：增加低PEG行业的配置，降低高PEG行业的配置。

```python
#作图，判断各行业PEG是否在PEG均值的上下一倍标准差以内
data_sorted = data_processed.sort('PEG')
fig = plt.figure(figsize=(12, 6))
ax = fig.add_subplot(111)
x = range(28)
plt.plot(x, data_sorted['PEG'],'bx')
x_labels = data_sorted.index
s = ax.set_xticks(x)
s = ax.set_xticklabels(x_labels, minor=False, fontproperties=font, fontsize=10, rotation=45)
ax.plot([data_sorted['PEG'].mean()-data_sorted['PEG'].std()]*len(data_sorted),'b--', label=u'+1标准差')
ax.plot([data_sorted['PEG'].mean()]*len(data_sorted),'r',lw=2,label=u'PEG均值')  
ax.plot([data_sorted['PEG'].mean()+data_sorted['PEG'].std()]*len(data_sorted),'r--',label=u'-1标准差') 
ax.legend(fontsize='x-small', prop=font)
plt.ylabel(u'PEG')
plt.xlabel(u'行业', fontproperties=font)
```

## 五、	PEG选股策略的短期表现
---
* 基于上述分析，我们设计了PEG选股策略：选取每月PEG低于合理中枢一倍标准差的行业的所有沪深A股，以等权方式构建单一做多投资组合。

```python
raw_data = pd.read_csv('data.csv', encoding='gbk')

#读取申万一级行业分类中各行业对应的股票
industries = [u'采掘', u'传媒', u'电气设备', u'电子', u'房地产', u'纺织服装', u'非银金融', u'钢铁', u'公用事业', u'国防军工', u'化工', u'机械设备', u'计算机', u'家用电器', u'建筑材料', u'建筑装饰', u'交通运输', u'农林牧渔', u'汽车', u'轻工制造', u'商业贸易', u'食品饮料', u'通信', u'休闲服务', u'医药生物', u'银行', u'有色金属', u'综合']

stocks = DataAPI.EquIndustryGet(industryVersionCD=u"010303")[['secShortName', 'industryName1', 'intoDate']].drop_duplicates(['secShortName', 'industryName1'])
stocks['industryName1'] = stocks['industryName1'].apply(lambda x: x.decode('utf-8')) 
stocks['secShortName'] = stocks['secShortName'].apply(lambda x: x.decode('utf-8')) 
stocks['intoDate'] = [time.strftime("%Y-%m-%d %H:%M:%S", time.strptime(stocks['intoDate'].iloc[i], "%Y-%m-%d")) for i in range(0,len(stocks))]

#只考虑申万一级行业，删除其它行业分类数据
ticker = raw_data[raw_data['industry'] == industries[0]]
for i in range(1,28):
    ticker = pd.concat((ticker, raw_data[raw_data['industry'] == industries[i]]))
ticker = ticker.drop_duplicates(['ticker'])[['ticker','industry']]

stocks = pd.merge(stocks, ticker, left_on='industryName1', right_on='industry')

data = raw_data[raw_data['ticker'] == ticker.iloc[0,0]]
for i in range(1,28):
    data = pd.concat((data, raw_data[raw_data['ticker'] == ticker.iloc[i,0]]))

data = data.drop_duplicates(['ticker','conDate','rptDate'])
data = data[['ticker', 'industry','conDate','rptDate','C4','C5','C7','C12','CB','CPB']]
data.columns = ['Code', 'Industry','Date','Year','Earning','PE','Growth','ROE','BV','PB']

#将对应于不同交易日的Long Dataframe转化为Short Dataframe形式，并将各个交易日PEG低于均值下一倍标准差的股票合并
dates = data.drop_duplicates('Date')['Date'].values
data_all = pd.DataFrame([])
stocks_all = pd.DataFrame([])

for date in dates:
    data_single = data[data['Date']==date]
    pivoted = data_single.pivot(index='Industry', columns='Year')
    earning = pivoted['Growth']
    earning.columns = ['Growth rate 0', 'Growth rate 1','Growth rate 2','Growth rate 3']
    roe = pivoted['ROE']
    roe.columns = ['ROE 0', 'ROE 1','ROE 2','ROE 3']
    
    year = np.sort(data_single.drop_duplicates(['Year'])['Year'].values)

    data_single = data_single.set_index('Industry')
    data_processed = pd.merge(data_single[data_single['Year'] == year[1]],earning,left_index=True, right_index=True)
    data_processed = pd.merge(data_processed, roe,left_index=True, right_index=True)
    data_processed['Compound growth rate'] = [np.sqrt((data_processed['Growth rate 2'][i]+1)*(data_processed['Growth rate 3'][i]+1))-1 for i in range(0,len(data_processed))]
    data_processed['PEG'] = data_processed['PE']/data_processed['Compound growth rate']/100
    data_processed['PB/ROE'] = data_processed['PB']/data_processed['ROE']
    data_processed = data_processed[(data_processed['PEG']>0) & (data_processed['PEG']<5) & (data_processed['Growth rate 0']<5) & (data_processed['Growth rate 0']>-5)]
    
    data_all = pd.concat((data_all, data_processed))
    
    data_processed = data_processed.sort('PEG')
        
    data_processed = data_processed[data_processed['PEG']<=data_processed['PEG'].mean()-data_processed['PEG'].std()]
        
    stocks_selected = pd.merge(data_processed, stocks, left_on='Code', right_on='ticker')
    stocks_selected = stocks_selected[stocks_selected['intoDate']<=stocks_selected['Date']][['Date', 'industry','secShortName']]
    stocks_all = pd.concat((stocks_all, stocks_selected))
    
stocks_all.columns = ['Date', 'Industry', 'Stock']
```

```python
#将被选取股票合并后生成的Dataframe作为交易信号
signal_monthly_df = stocks_all
signal_monthly_df['Date'] = signal_monthly_df['Date'].apply(lambda x: time.strftime("%Y/%m/%d", time.strptime(x, "%Y-%m-%d %H:%M:%S")))

#获取各股票对应的代码（只保留沪深两市A股）
stocks_all = DataAPI.SecIDGet(assetClass='E')
stocks_all = stocks_all[(stocks_all['exchangeCD']=='XSHE') | (stocks_all['exchangeCD']=='XSHG')]
stocks_all['secShortName'] = stocks_all['secShortName'].apply(lambda x:x.decode('utf-8'))
stocks_all = pd.merge(stocks_all, signal_monthly_df, left_on='secShortName', right_on='Stock')
signal_monthly_df = stocks_all[['Date', 'Industry', 'Stock']]
signal_monthly_df['Ticker'] = stocks_all['secID']
signal_monthly_df['Head_ticker'] = signal_monthly_df['Ticker'].apply(lambda x: x[:3])
signal_monthly_df = signal_monthly_df[(signal_monthly_df['Head_ticker']=='600') | (signal_monthly_df['Head_ticker']=='603') | (signal_monthly_df['Head_ticker']=='600') | (signal_monthly_df['Head_ticker']=='300') | (signal_monthly_df['Head_ticker']=='002')]

#获取回测时的调仓日期
monthly_exchange_date_list = signal_monthly_df['Date'].unique()
```

## 六、PEG选股策略的长期表现
---
* 以上回测仅考虑了2016年以来不足两年的回测区间。接下来，我们将对2010年以来将近8年的区间进行回测，以探究PEG选股策略是否长期有效。下图展示了2010年1月至2017年8月的回测结果。
* 从长期来看，PEG选股策略的表现同样超过中证500，但表现差异弱于长期，且夏普比率较低，最大回撤超过50%。这说明PEG选股策略受市场环境的影响，在某些特定的市场环境下更为有效。
* 于是，我们对各行业PEG的历史数据进行了分析，发现多年以来，各行业的PEG基本都在一个稳定的区间内来回波动，其中钢铁、房地产、汽车、建筑装饰、化工等大盘股集中的行业的PEG一直处于所有行业的下游，而休闲服务、传媒、商业贸易等小盘股集中的行业往往处于上游。因此，依据PEG选股通常会选出盘子较大的工业股，而非中小板的轻资产股票。也就是说，PEG选股本质上等同于规模因子选股，但选择的是规模大的股票，而非规模小的。2016年以前，我国股市小盘股表现往往远超大盘股，这使得PEG选股策略表现一般。而2015年下半年股灾过后，虽然大盘指数逐渐恢复，但小盘股的表现却不尽如人意。因此，PEG选股策略在当今的市场环境下是相对有效的，但这一策略受到市场环境的诸多限制。
