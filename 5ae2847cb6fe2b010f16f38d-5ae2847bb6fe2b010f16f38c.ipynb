1.熟悉三张表中的勾稽关系，以下问题都指同一家公司
a. 期末的资产负债表显示该公司，有4亿总资产，2.5亿股东权益，问该公司总负债是多少
b. 利润表中显示该公司本期有3000万净利润，1.75亿总费用，问该公司营收是多少
c. 期初公司有2.3亿股东权益，本期公司净支付给股东1000万，假设公司不进行股票回购、增发股份，问公司发放红利是多少
提示: Dividends=Net Payout + Share Repurchase - Share Issued



```python
# a
total_assets = 4
equity_share = 2.5
liability = total_assets - equity_share
print(u'负债:{0:.2f}亿'.format(liability))
# b
net_income = 0.3
expense = 1.75
revenue = net_income + expense
print(u'营收:{0:.2f}亿'.format(revenue))
# c
equity_share_begin = 2.3
net_payout = 0.1
share_repurchase = 0.0
share_issued = 0.0
dividends = net_payout + share_repurchase - share_issued
print(u'发放分红:{0:.2f}亿'.format(dividends))
```

2.银行储蓄的融资分析
公司年初在银行存了100万，存储利率年化5%，年底时公司分红5万，填如下表格
a. 资产负债表，现金资产:______，股东权益:______
b. 利润表，收入:______，成本:______，利润:______
c. 现金流量表，经营活动净额:______，投资活动净额:______，融资活动净额:______



```python
cash_begin = 100
interest = 0.05
dividend = 5
revenue = cash_begin * interest
# 利润表
expense = 0.0
earnings = revenue - expense
print(u'-----利润表-----')
print(u'收入:{0:.2f}万'.format(revenue))
print(u'成本:{0:.2f}万'.format(expense))
print(u'利润:{0:.2f}万'.format(earnings))
# 现金流量表
operate_cash_flow = revenue
investment_cash_flow = 0.0
finance_cash_flow = -dividend
print(u'-----现金流量表-----')
print(u'经营活动净额:{0:.2f}万'.format(operate_cash_flow))
print(u'投资活动净额:{0:.2f}万'.format(investment_cash_flow))
print(u'融资活动净额:{0:.2f}万'.format(finance_cash_flow))
# 资产负债表
cash_end = cash_begin + revenue - dividend
liability = 0.0
share_equity = cash_end - liability
print(u'-----资产负债表-----')
print(u'资产(现金):{0:.2f}万'.format(cash_end))
print(u'所有者权益:{0:.2f}万'.format(share_equity))
```

3.简单的折现现金流估值
公司当前净债务7.59亿，假设未来三年企业经营活动现金流为14.5亿、15.76亿、17.18亿；未来三年投资现金10.2亿、11.24亿、12亿；永续增长率为4%，平均资本成本为10%，请回答下列问题
a. 计算当前公司企业价值
b. 计算当前公司股权价值
hint: 企业价值包含股权价值+债券价值两部分

```python
import pandas as pd
import numpy as np

continuedGrowth = 0.04
wacc = 0.1
debtValue = 7.59

df = pd.DataFrame([[14.5, 10.2], [15.76, 11.24], [17.18, 12]], columns=['operateCashFlow', 'investCashFlow'])
df['freeCashFlow'] = df['operateCashFlow'] - df['investCashFlow']
df['discountRatio'] = wacc
df['discountRatio'] = (1.0 + df['discountRatio']).rolling(window=len(df['discountRatio'].index), min_periods=0, center=False).apply(lambda x: np.prod(x))
df['PVCashFlow'] = df['freeCashFlow'] / df['discountRatio']
print(df.to_html())
continuedValue = df.iloc[-1]['freeCashFlow'] * (1 + continuedGrowth) / (wacc - continuedGrowth)
pvContinuedValue = continuedValue / df.iloc[-1]['discountRatio']
print(u'永续现金流: {0:.2f}亿, 折现后永续现金流: {1:.2f}亿'.format(continuedValue, pvContinuedValue))

enterpriseValue = np.sum(df['PVCashFlow']) + pvContinuedValue
equityValue = enterpriseValue - debtValue

print(u'企业价值:{0:.2f}亿'.format(enterpriseValue))
print(u'股权价值:{0:.2f}亿'.format(equityValue))
```

```python
# step 3) 财务分析
def load_fiscal_data(ticker='000002', end_date='2017-09-30'):
    """
    参考DataAPI文档: https://uqer.datayes.com/data/
    """
    bs = DataAPI.FdmtBSGet(ticker=ticker,secID=u"",reportType=u"",endDate=u"",beginDate=u"2013-09-30",
                                    publishDateEnd=u"",publishDateBegin=u"",\
                                    endDateRep="",beginDateRep="",beginYear="",endYear="",fiscalPeriod="",\
                                    field=u"ticker,endDate,publishDate,fiscalPeriod,TAssets,inventories,AR,TCA,TCL,fixedAssets,TLiab,TShEquity",pandas="1")
    bs = bs.sort_values(by=['endDate', 'publishDate', 'fiscalPeriod'], ascending=[True, True, False]).drop_duplicates(subset=['endDate'], keep='first')
    income = DataAPI.FdmtISGet(ticker=ticker,secID=u"",reportType=u"",endDate=u"",beginDate=u"2013-09-30", \
                                    publishDateEnd=u"",publishDateBegin=u"",\
                                    endDateRep="",beginDateRep="",beginYear="",endYear="",fiscalPeriod="", \
                               field=u"ticker,secShortName,endDate,publishDate,fiscalPeriod,tRevenue,TCogs,COGS,TProfit,NIncome,incomeTax,adminExp",pandas="1")
    income = income.sort_values(by=['endDate', 'publishDate', 'fiscalPeriod'], ascending=[True, True, False]).drop_duplicates(subset=['endDate'], keep='first')
    bs = bs.merge(income, left_on=['ticker', 'endDate', 'publishDate', 'fiscalPeriod'], right_on=['ticker', 'endDate', 'publishDate', 'fiscalPeriod'], how='inner')
    bs['fiscalPeriod'] = bs['fiscalPeriod'].astype(int)
    # 注意这里我们将nan值填为0
    bs = bs.fillna(0.0)
    return bs

df = load_fiscal_data(ticker='600276', end_date='2017-12-31')
df = df[df['fiscalPeriod']==12].copy(deep=True).set_index('endDate', drop=False)
df['researchRatio'] = df['adminExp']/df['tRevenue']

fig, ax = plt.subplots()
ax2 = ax.twinx()
df['tRevenue'].plot.bar(label='totalRevenue', ax=ax, color='b')
df['researchRatio'].plot(label='researchRatio', ax=ax2, color='r')
ax.set_title(u"恒瑞医药:研发费用投入vs总收入", fontproperties=font, fontsize=16)
ax.legend(loc=2)
ax2.legend()

# step 4) 横向比较
dic = {"600276":u"恒瑞医药","600521":u"华海药业","300630":u"普利制药","600196":u"复星医药","300347":u"泰格医药"}
df = DataAPI.MktStockFactorsOneDayProGet(tradeDate=u"20171229",secID=u"",ticker=",".join([i for i in dic.keys()]),field=u"ticker,ForwardPE",pandas="1")
fig, ax = plt.subplots()
df['ForwardPE'].plot.bar(ax=ax)
ax.set_xticklabels([dic[df.iloc[i]['ticker']] for i in df.index], fontproperties=font, fontsize=16)
ax.set_title(u"同行业PE对比", fontproperties=font, fontsize=16)
_ = plt.xticks(rotation=45)
```
