```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt

begin_date = '20090101'
end_date = '20170415'
universe = set_universe('A')


'''————————————————————————————————————————————分割线————————————————————————————————————————————————————————————————'''


def get_industry(universe):
    match_data = DataAPI.IndustryGet(industryVersionCD=u"010303",industryLevel=u"1",isNew=u"1",field=u"industryID,indexSymbol").set_index('industryID')                     # 得到industryID和indexSymbol一一对应的dataframe，因为只能用后者查到股票对应行业信息
    all_data = pd.DataFrame()     #初始all_data数据表
    for stk in universe:          #实现各个股票有着对应的industryID和indexSymbol
        tmp_data=DataAPI.EquIndustryGet(industryVersionCD=u"010303",secID=stk,field=u"ticker,industryID",pandas="1").ix[[0]]
        tmp_data.industryID = int(round(tmp_data.industryID.values/10000))
        for x in match_data.index.values:
            if x == tmp_data.industryID.values:
                tmp_data['indexSymbol'] = match_data.ix[x,0]
        all_data = [all_data,tmp_data]    #此为整合每个tmp_data
        all_data = pd.concat(all_data)    
    return all_data
    
    
'''————————————————————————————————————————————分割线————————————————————————————————————————————————————————————————'''
    
    
def get_datelist(begin_date,end_date):
    date_data=DataAPI.TradeCalGet(exchangeCD=u"XSHG",beginDate=begin_date,endDate=end_date,field=['calendarDate','isOpen'])
    date_data=date_data[date_data['isOpen'] == 1]
    #date= map(lambda x: x[0:4]+x[5:7]+x[8:10], date_data['calendarDate'].values.tolist())
    date_data=date_data['calendarDate'].values.tolist()
    return date_data               #得到交易日

'''————————————————————————————————————————————分割线————————————————————————————————————————————————————————————————'''

#先运行get_industry(),得到match_data后保存
# match_data = get_industry(universe)
# match_data.to_csv('match_data.csv')


match_data = pd.read_csv('match_data.csv')
tmp = [str(x) for x in match_data.ticker]
tmp = [x.zfill(6) for x in tmp]
match_data.ticker = tmp
match_data = match_data[match_data.columns[1:]].set_index('ticker')     #完成从读取到整理match_data
match_data = match_data.dropna()

date_list = get_datelist(begin_date,end_date)
#完成运行get_R()的准备数据。


'''————————————————————————————————————————————分割线————————————————————————————————————————————————————————————————'''


def get_R(begin_date,end_date,match_data,date_list):
    all_data = pd.DataFrame(index=range(len(date_list)),columns=['tradeDate'],data=date_list)    
    for x in match_data.index.values:
        int_x = int(x)
        mkt1 = int(match_data.ix[x,'indexSymbol'])
        if int_x >= 600000:
            mkt2 = '000001'
        if int_x <=1000:
            mkt2 = '399001'
        if int_x >= 300000 and int_x <=400000:
            mkt2 = '399006'
        tmp_stk = DataAPI.MktEqudGet(beginDate=begin_date,endDate=end_date,ticker=x,field=u"tradeDate,chgPct",pandas="1")
        tmp_mkt1 = DataAPI.MktIdxdGet(beginDate=begin_date,endDate=end_date,ticker=str(mkt1),exchangeCD=u"XSHE,XSHG",field=u"tradeDate,CHGPct",pandas="1")
        tmp_mkt2 = DataAPI.MktIdxdGet(beginDate=begin_date,endDate=end_date,ticker=mkt2,exchangeCD=u"XSHE,XSHG",field=u"tradeDate,CHGPct",pandas="1")
        tmp_mkt1.columns = ['tradeDate','mkt1_P']
        tmp_mkt2.columns = ['tradeDate','mkt2_P']
        tmp = pd.merge(tmp_stk,tmp_mkt1)
        tmp = pd.merge(tmp,tmp_mkt2)    #完成计算相关系数所需的数据的合并,到此步tmp没问题
        tmp.chgPct = tmp.chgPct*100
        tmp.mkt1_P = tmp.mkt1_P*100
        tmp.mkt2_P = tmp.mkt2_P*100
        tmp = tmp.dropna()
        for i in range(len(tmp.index)-20):
            tmp_x = tmp.ix[i:i+20,['mkt1_P','mkt2_P']]
            tmp_y = tmp.ix[i:i+20,['chgPct']]
            tmp_x = sm.add_constant(tmp_x)
            est=sm.OLS(tmp_y,tmp_x).fit()
            tmp_R = est.rsquared      
            tmp.loc[i+20,'R'] = tmp_R
        tmp = tmp.ix[:,['tradeDate','R']]
        tmp.columns = ['tradeDate',x]   
        if all_data.empty:
            all_data = tmp
        else:all_data = pd.merge(left=all_data,right=tmp,how='left')
    return all_data        #得到A股市场全部股票的R平方


'''————————————————————————————————————————————分割线————————————————————————————————————————————————————————————————'''

        
#R = get_R(begin_date,end_date,match_data,date_list)
#R.to_csv('all_A_R_cp.csv')
R = pd.read_csv('all_A_R_cp.csv')
R = R.ix[:,1:].set_index('tradeDate')    #完成全A股09到14年R平方的计算


print match_data.head()
print date_list
print R




```
