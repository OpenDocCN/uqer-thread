2. $y-1\geqslant0$

求解$f(x,y)$的最大值

作业1

```python
# 这是一个带约束的优化问题，按照教程中的方法，使用SLSQP（Sequential Least SQuares Programming optimization algorithm）方法进行约束问题的求解

from scipy import optimize

# 定义函数，因为求最大值，函数取反，转为求最小值以便优化
def f(x):
    return - (2 * x[0] * x[1] + 2 * x[0] - x[0]**2 - 2 * x[1]**2)

# 约束条件
con = (
    {'type': 'eq', 'fun': lambda x: 2 * x[0] - x[1]},
    {'type': 'ineq', 'fun': lambda x: x[1] - 1}
)

# 求解
x0 = [0, 0]  #初始值
x_constrain = optimize.minimize(f, x0, method='SLSQP', constraints=con).x
print('带约束的极值点：%s' % x_constrain)
```

```python
# 解完题，顺便看看函数长啥样

# 设定x, y范围，求出z（取反还原出原函数）
x = np.arange(-4, 4, 0.1)
y = np.arange(-4, 4, 0.1)
x, y = np.meshgrid(x, y)
z = -f([x, y])

# 3D作图
ax = plt.subplot(projection='3d')
ax.plot_surface(x, y, z, cmap=cm.coolwarm, alpha=0.8)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.view_init(azim=-20)  #调整视角
# 加入等高线图，方便观察极值点
ax.contourf(x, y, z, zdir='z', offset=-120, cmap=cm.coolwarm, alpha=0.8, antialiased=True)
# 极值点
x = optimize.minimize(f, x0, method='SLSQP').x
print('不带约束的极值点：%s' % x)
ax.scatter(x[0], x[1], -120, c='black') #不带约束，黑色
ax.scatter(x_constrain[0], x_constrain[1], -120) #带约束，蓝色
```
