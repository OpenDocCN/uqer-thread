三行python 代码做一个动量策略

```python
import numpy as np
df=DataAPI.MktIdxdGet(indexID=u"000300.ZICN",ticker=u"",tradeDate=u"",beginDate=u"20100101",endDate=u"20200707",exchangeCD=u"XSHE,XSHG",field=u"indexID,tradeDate,preCloseIndex,closeIndex",
                      pandas="1")
df['000300.ZICN']=df.apply(lambda x:(x.closeIndex-x.preCloseIndex)/x.preCloseIndex, axis = 1)
df=df[1:]
#print df
code="000300.ZICN"
#numpy.sign可以向量化的判断收益率为正或负。——其实收益率就是1天的动量。
df['position'] = np.sign(df[code])
print df.head()
#下面我们做一个最简单的动量策略，就是昨天收益率（动量）为正当天持仓; 如果昨天为负，则当天空仓，一行代码就够了：
df['strategy'] = df['position'].shift(1) * df[code]
print df.head()
#策略收益图 cumsum()是连乘的意思，等于（1+rate).cumprod()。
(df[[code, 'strategy']]+1).cumprod().plot(figsize=(10, 6))

#创业板指数的13天动量策略
code="399006.ZICN"
df1=DataAPI.MktIdxdGet(indexID=code,beginDate=u"20110101",endDate=u"20200707",exchangeCD=u"XSHE,XSHG",field=u"indexID,tradeDate,preCloseIndex,closeIndex",pandas="1")
df1['399006.ZICN']=df1.apply(lambda x:(x.closeIndex-x.preCloseIndex)/x.preCloseIndex, axis = 1)
df1['position'] = np.sign(df1[code])
df1=df1[1:]
df=df1
print "df=",df.head()
df['equity'] = (1+df[code]).cumprod()
print df.head()
N = 13
df['equity_{}'.format(N)] = df['equity'].shift(N)
print "2--df=",df
df['momentum'] = df['equity']/df['equity_{}'.format(N)]-1
df['position'] = np.sign(df['momentum'])
df['strategy'] = df['position'].shift(1) * df[code]
print "累计收益：\n",(df[[code, 'strategy']]+1).cumprod().tail()
(df[[code, 'strategy']]+1).cumprod().plot(figsize=(10, 6))
```
