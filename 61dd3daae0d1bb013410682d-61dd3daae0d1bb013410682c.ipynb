```python
import pandas as pd
import numpy as np
import datetime
import matplotlib.pyplot as plt
```

```python
factor = pd.read_csv('factor17.csv',header=0,index_col=0)
factor.index = pd.to_datetime(factor.index,format='%Y-%m-%d')
factor.head()
```

```python
#每天买入的股票列表
factor_frame_list = []
for i in range(len(factor.index)):
    factor_frame = factor.iloc[i,:].sort_values(ascending = False).head(5).index
    factor_frame_list.append(factor_frame)
```

```python
#计算持有十天的收益
rtrn = factor.copy()
for i in range(len(rtrn.index)-10):
    rtrn.iloc[i,:] = factor.iloc[i+10,:] - factor.iloc[i,:]
rtrn.head()
```

```python
results = pd.DataFrame(index = factor.index,columns = factor.columns)
for i in range(len(factor.index)):
    for j in range(5):
        results[factor_frame_list[i][j]][i] = rtrn[factor_frame_list[i][j]][i]
results.to_csv('results.csv')
```

```python
results['sum'] = results.apply(lambda x: x.sum(), axis=1)
plt.plot(results.index,results['sum'])
```
