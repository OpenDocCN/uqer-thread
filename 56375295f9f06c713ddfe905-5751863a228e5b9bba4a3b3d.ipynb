* `refresh_date 函数  ：` 返回每个月最后一个交易日列表　

```python
# start : 回测开始时间　end : 回测结束时间
def refresh_date(start , end) : 
    # DataAPI.TradeCalGet 接口 ： 获取交易日，将每月最后一个交易日放入date_list列表，月末调仓
    data = DataAPI.TradeCalGet(exchangeCD=u"XSHG",beginDate = start ,endDate= end ,field=['calendarDate','isMonthEnd'] ,pandas="1")
    data = data[data['isMonthEnd'] == 1]
    date_list = data['calendarDate'].values.tolist()
    return date_list
```

* `Low_PB 函数  ：`　返回PB小于2的个股

```python
# universe : 股票池　trade_time : 参考时间 
def Low_PB(universe,trade_time) :    
    buylist = []
    for s in universe : 
        try : 
            if DataAPI.MktStockFactorsOneDayGet(tradeDate = trade_time ,secID= s , field=u"PB",pandas="1")['PB'][0] <= 2 : 
                buylist.append(s)
        except : 
            pass         
    return buylist
```

* `Low_DebtsAssetRatio 函数  ：`　返回资产负债比小于市场平均值的个股

```python
def Low_DebtsAssetRatio(universe,trade_time) :        
    data = DataFrame()
    #首先我们需要计算A股资产负债比的中位数，作为市场平均参考值  
    for s in range(len(set_universe('A'))/300 + 1) :        
        if s == len(set_universe('A'))/300 :
            temp_list = set_universe('A')[s*300:]
        else :
            temp_list = set_universe('A')[s*300:(s+1)*300]              
        if not len(temp_list) == 0 : 
            data_temp = DataAPI.MktStockFactorsOneDayGet(tradeDate = trade_time,secID=temp_list,field=u"secID,DebtsAssetRatio",pandas="1")   
            data = pd.concat([data,data_temp])             
    data.set_index('secID',inplace=True)
    ans = data.median()   
    
    buylist = []
    for s in universe : 
        if data['DebtsAssetRatio'][s] <= ans['DebtsAssetRatio'] : 
            buylist.append(s)
    
    return buylist
```

> 2.   负债比例低于市场平均值。

* `SE_change 函数 ： `　返回公司经营阶层持股稳定的个股

```python
# universe : 股票池　trade_time : 参考时间 
def SE_change(universe,short_Date,long_Date,end_Date) :            
    # 　购买（年内）高管增持的股票，剔除短期内(调仓周期内)高管减持的股票        
    buylist = []
    for s in universe : 
        try : 
            temp = DataAPI.JY.EquMsChanJYGet(secID = s ,field=u"changeDate,sharesChange",pandas="1")
            temp['changeDate'] = pd.to_datetime(temp['changeDate'])
            temp = temp[temp['changeDate'] > pd.to_datetime(long_Date)]
            temp = temp[temp['changeDate'] < pd.to_datetime(end_Date)]
            if np.sum(temp['sharesChange']) > 0 : 
                buylist.append(s)
            temp = temp[temp['changeDate'] > pd.to_datetime(short_Date)]
            #短期内高管减持剔除
            if np.sum(temp['sharesChange']) < 0 : 
                buylist.remove(s)
        except : 
            pass                        
    return buylist
```
