```python
def preceding_date(date):
    cal = DataAPI.TradeCalGet(exchangeCD=u"XSHG", beginDate='20110101', endDate=date, field=[u"calendarDate", u"isOpen"], pandas="1")
    cal = cal[cal['isOpen'] == 1]
    date = cal['calendarDate'].values[-2].replace('-', '')
    return date

def duotou_5_10(date, stockList, precedingDate=True):
    if precedingDate:
        date = preceding_date(date)
    duotou = {}
    if stockList is None or len(stockList) == 0:
        return duotou
    kLine = DataAPI.MktStockFactorsOneDayGet(tradeDate=date, secID=stockList, field=[u"secID", u"MA5", u"MA10"], pandas="1")
    kLine = kLine.dropna()
    for stock, ma5, ma10 in zip(kLine['secID'].values, kLine['MA5'].values, kLine['MA10'].values):
        if ma5 > ma10:
            duotou[stock] = True
        else:
            duotou[stock] = False
    return duotou

def spreadRateByIntrinsicValue(context, overflow=0.0, precedingDate=True):
    stock_list = context.get_universe()
    current_date = context.current_date
    date = current_date.strftime('%Y%m%d')
    if precedingDate:
        date = preceding_date(date)
    eq_EPS_EGRO = DataAPI.MktStockFactorsOneDayGet(tradeDate=date, secID=stock_list, field=[u"secID", u"EPS", u"EGRO"], pandas="1")
    eq_EPS_EGRO['Value'] = eq_EPS_EGRO['EPS'] * ( 8.5 + 2 * eq_EPS_EGRO['EGRO'] / 5)
    eq_EPS_EGRO = eq_EPS_EGRO.dropna()
    eq_price = DataAPI.MktEqudAdjGet(secID=stock_list, tradeDate=date, beginDate=u"", field=[u"secID", u"closePrice"], pandas="1")
    eq_price = eq_price.dropna()
    eq_EPS_EGRO = eq_price.merge(eq_EPS_EGRO, on='secID', how='inner')
    spread_rate = []
    for stock, intrinsic_value, reference_price in zip(eq_EPS_EGRO['secID'].values, eq_EPS_EGRO['Value'].values, eq_EPS_EGRO['closePrice']):
        intrinsic_value = intrinsic_value * (1 + overflow)
        if reference_price > 0 and reference_price < intrinsic_value:
            spread_rate.append((stock, (intrinsic_value - reference_price) / reference_price))
    return sorted(spread_rate, key=lambda k: k[-1], reverse=True)

'''
判断是否为底部形态，判断标准为股价相对于近期高点大幅下跌超过downPercent，并且收盘价在五日线十日线之下，并且收红或者收星，跌幅小于7%
'''
def isButtom(date, stockList, precedingDate=True, downPercent=0.3):
    cal = DataAPI.TradeCalGet(exchangeCD=u"XSHG", beginDate='20110101', endDate=date, field=u"prevTradeDate", pandas="1")
    daysAhead = cal['prevTradeDate'].values[-20].replace('-', '')
    if precedingDate:        
        date = cal['prevTradeDate'].values[-1].replace('-', '')
    rs = {}
    if stockList is None or len(stockList) == 0:
        return rs
    dayInfo = DataAPI.MktEqudAdjGet(secID=stockList, beginDate=daysAhead, endDate=date ,field=[u"secID", u"openPrice", u"closePrice", u"preClosePrice"], pandas="1")
    dayInfo.dropna()
    for stock in stockList:
        stockDayInfo = dayInfo[dayInfo['secID'] == stock]
        closePrices = stockDayInfo['closePrice'].values
        ma5 = np.mean(closePrices[-5:])
        ma10 = np.mean(closePrices[-10:])
        closePrice = closePrices[-1]
        maxClosePrice = np.max(closePrices)
        openPrice = stockDayInfo['openPrice'].values[-1]
        preClosePrice = stockDayInfo['preClosePrice'].values[-1]
        if (maxClosePrice-closePrice) / maxClosePrice > downPercent and closePrice < ma5 and ma5 < ma10 and (closePrice > openPrice or abs(closePrice - openPrice) / openPrice < 0.02) and abs(closePrice - preClosePrice) / preClosePrice < 0.07:
            rs[stock] = True
        else:
            rs[stock] = False
    return rs
```

```python
def handle_data(context):                                    # 每个交易日的买入卖出指令
    global max_percent_of_a_stock                            # global是定义全局变量
    buylist = []
    selist = []                                              # 定义了2个列表
    current_date = context.current_date
    current_date = current_date.strftime('%Y%m%d')           # 获取当前日期，并且转化成字符串
    stock_account = context.get_account('stock_account')     # 获取当前账户

    overflow = 0.15                                          # 根据情况给予一定的溢价（例如0.1)或者折价(例如-0.1)，也可以根据市场风险程度进行动态调节（此处读者可以自行发挥）

    spread_rate = dict(spreadRateByIntrinsicValue(context, overflow=overflow, precedingDate=True))  # 调用了spreadratebyintrinsicvalue这个函数，不清楚功能的看上一篇，返回值是一个字典
            
    referencePortfolioValue = stock_account.portfolio_value  # account.pv 参考价值，就是回测时策略的净值

    # 获取用来计算多头形态的股票列表
    stock_set_for_duotou = []
    stock_set_for_duotou.extend([stock for stock in stock_account.get_positions().keys() if stock_account.get_position(stock).available_amount > 0])# 可卖证券头寸的key部分（股票代码）
    stock_set_for_duotou.extend(spread_rate.keys())          # 计算过溢价比例的key部分（股票代码）
    stock_set_for_duotou = list(set(stock_set_for_duotou))   # set删除重复部分
    
    duotou_5_10_Map = duotou_5_10(current_date, stock_set_for_duotou, precedingDate=True)  # doutou_5_10这个自定义函数出马了
    isButtom_Map = isButtom(current_date, stock_set_for_duotou, precedingDate=True, downPercent=0.3)  # isbuttom这个自定义函数出马了,可以查看前面的
    
    for stock in [stock for stock in stock_account.get_positions().keys() if stock_account.get_position(stock).available_amount > 0]:
        if stock not in spread_rate and not duotou_5_10_Map.get(stock, False):  # 如果可卖的股票不在spread_rate并且不在duotou_5_10_Map里
            selist.append(stock)                             # 添加该代码到selist中。selist应该就是卖出列表
    
    for stock in selist:
        sell_value = context.current_price(stock) * stock_account.get_position(stock).available_amount  # 遍历selist中的股票，selist中的股票的昨日收盘价*可卖数量赋值给sell_value
        stock_account.order_to(stock, 0)                     # 卖出selist中的股票
            
    for stock in spread_rate.keys():
        if stock not in stock_account.get_positions(exclude_halt=True).keys():  # 遍历spread_rate中的股票，如果该股票没持仓
            buylist.append(stock)                            # 吧股票代码添加到buylist中
    
    for stock in buylist:
        # 满足以下条件之一买入：（1）5日线在10日线之上；（2）出现底部特征
        if duotou_5_10_Map.get(stock, False) or isButtom_Map.get(stock, False):
            buy_value = min(referencePortfolioValue * max_percent_of_a_stock, stock_account.cash / len(buylist))  # 净值*单只股票持仓最大仓位 和 现金/购买列表 中的最小值赋值给buy_value
            if buy_value / referencePortfolioValue >= 0.0001:  # 如果buy_value/净值>=0.0001
                stock_account.order_pct(stock, buy_value / referencePortfolioValue)  # 买入股票，仓位比例是buy_value/净值
```

### 这段代码其实写的比较繁琐，下面指出哪几处可以简化

1、precedingdate这个函数其实是非必要的，因为优矿提供了前一个工作日的简便写法：context.previous_date，直接调用就是前一个工作日的日期

2、sell_value这个函数在里面其实是没有起到任何作用的，很可能是作者调试时添加，之后忘了删除了

`最重要的一点：`

3、这个策略其实是调用了未来函数的，

因为在计算估值的时候用到了这样的语句`eq_price = DataAPI.MktEqudAdjGet(secID=stock_list, tradeDate=date, beginDate=u"", field=[u"secID", u"closePrice"], pandas="1")`调用的是前复权的价格，也就是说这里是判断前复权价格是不是小于估值，

如果调用了当时的实际价格，这个策略的收益是远远低于预期的。所以现在进行回测会和回测曲线完全不一样。
