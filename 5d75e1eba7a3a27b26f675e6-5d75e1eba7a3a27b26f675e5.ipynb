```python
Open = np.array(df['openPrice'])
High = np.array(df['highestPrice'])
Low = np.array(df['lowestPrice'])
Close = np.array(df['closePrice'])
```

模式识别
返回结果：0,100,-100

函数名：CDLMORNINGSTAR
名称：Morning Star 晨星
简介：三日K线模式，下跌趋势，第一日阴线，
第二日价格振幅较小，第三天阳线，预示底部反转。

```python
df['MorningStar']=talib.CDLMORNINGSTAR(Open, High, Low, Close,penetration=0)
```

```python
df[df['MorningStar']==100]
```

函数名：CDLSHOOTINGSTAR
名称：Shooting Star 射击之星
简介：一日K线模式，上影线至少为实体长度两倍，
没有下影线，预示着股价下跌

```python
df['ShootingStar']=talib.CDLSHOOTINGSTAR(Open, High, Low, Close)
df[df['ShootingStar']==-100]
```

```python
import matplotlib.finance as mpf
from matplotlib.pylab import date2num
import datetime
 
# 对tushare获取到的数据转换成candlestick_ohlc()方法可读取的格式
data_list = []
stockPrice = DataAPI.MktEqudGet(secID=u"",ticker=u"601318",tradeDate=u"",beginDate=u"20150901",endDate=u"20190901",isOpen="",field=u"tradeDate,openPrice,highestPrice,lowestPrice,closePrice",pandas="1").set_index('tradeDate')
for dates,row in stockPrice.iterrows():
    # 将时间转换为数字
    date_time = datetime.datetime.strptime(dates,'%Y-%m-%d')
    t = date2num(date_time)
    Open,High,Low,Close = row[:4]
    datas = (t,Open,High,Low,Close)
    data_list.append(datas)
# 创建子图
fig, ax = plt.subplots()
fig.subplots_adjust(bottom=0.2)
# 设置X轴刻度为日期时间
ax.xaxis_date()
plt.xticks(rotation=45)
plt.yticks()
plt.title("ticker:601318")
plt.ylabel("stockPrice")
mpf.candlestick_ohlc(ax,data_list,width=1.5,colorup='r',colordown='green')
plt.grid()
```
