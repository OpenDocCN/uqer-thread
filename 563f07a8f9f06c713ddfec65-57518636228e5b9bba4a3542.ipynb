二、 随便聊聊

# 2.1 发行和上市

企业的发行和上市，是截然不同又相辅相成的两件事。发行主要是一级市场的事，上市是二级市场的事。

证券发行和上市制度，主要由发行和上市审核制度、信息披露制度、发行定价制度三个方面构成。

在近来的IPO重启的新闻中，一定会涉及上面这几块内容。

从法律上来看，上市审核权和股票发行审核权是分离的。但因为交易所事实上是证监会的“派出机构”，所以按现状来说，不会出现证监会批准发行，但交易所驳回上市的情况。

所以，一般的文章中都是混用。

# 2.2 没有交易所的日子里

证券市场，一般并不是从集中交易的交易所开始，而是从OTC柜台交易开始的。中国的证券市场也一样。

中国最早的两个柜台交易股票是“小飞乐”和“延中”，也就是飞乐音响和延中实业。这两个股票在交易所出现后，就有了代号600651和600601。

从这个角度看，也可以看出股票发行和股票上市是分离的。

# 3.3 “老八股”和“老五股”阶段

终于等到上海交易所和深圳交易所先后成立。股票终于有了可以上市的地方。

交易所开始批准逐步批准股票上市。

最早上市的股票，当然是沪市的“老八股”和深市的“老五股”。这些股票现在是收藏爱好者千金难买的珍品。

老八股分别是延中实业(600601),上海真空电子(600602),飞乐音响(600651),爱使股份(600652),申华电工(600653),飞乐股份(600654),豫园商城(600655),凤凰化工(600656)

现在他们分别为方正科技(600601),广电电子(600602),飞乐音响(600651),游久游戏(600652),申华控股(600653),中安消(600654),豫园商城(600655),ST方源(600656)

小飞乐还是小飞乐，豫园商城还是豫园商城，都不知道变成了什么。游久游戏是什么鬼？

---------------------------------------

老五股分别是深发展(000001)、深万科(000002)、深金田(000003)、深安达(000004)、深原野(000005)。

他们分别为平安银行(000001)、万科A(000002)、PT金田A(000003)、国农科技(000004)、世纪星源(000005)

三、 历次IPO暂停和重启概览
=============

现在，就让我们回顾下历次暂停及重启后对A股影响。文中采用的各个阶段的划分，是按照网络上流行的方法来处理的，可能是以发行日期做的区分。

现在可以获得的数据，只有上市日期，所以筛选出的数据可能会有一些出入。因为是概要的描述，所以不做区分了。

## 第一阶段
1990年12月-1994年7月21日正常发行，1994年7月21日-1994年12月7日暂停发行。空窗期：五个月

```python
first_period = equity_in_china[equity_in_china.listDate<'1994-12-07']

first_period.to_csv('ipo_first_period.csv', encoding='GB18030')
print "第一阶段(1990年12月-1994年12月7日)。共发行股票：" + str(first_period.secID.count()) + "只"
first_period.head(10)
```

```python
index_first_period = shanghai_index[shanghai_index.tradeDate<'1994-12-07']
index_first_period.index = index_first_period.tradeDate
index_first_period['closeIndex'].plot(figsize=(15,8))
```

第一阶段，是所有九个阶段中最重要，也最具代表性的。

从图中，明显可以看到1992年11月左右，从386点，曲线突然上涨，直到1993年2月，达到最高点1558点。

而这时，政府主导的证券市场大扩容也开始了。为了完成国有企业“三年解困”的政治任务，大量国有企业采用股权融资的方式，上市获取资金，解决自身面临的问题。

从此，指数开始向下，开始了第一个熊市。这也是中国股市经历的第一个牛熊转换。

由于熊市，出现了股民自缢身亡的新闻。这个2015年的情况如此相似。

1994年7月30日，证监会发布《中国证监会与国务院有关部门就稳定和发展股市做出决策》，提出了三大救市政策。
* 今年内暂停新股发行和上市
* 严格控制上市公司配股规模
* 采取措施扩大入市资金范围

这是第一次暂停新股发行和上市。股价应声上涨。但没多久，恢复了IPO，股价又维持下跌。

---------------------------------------

从这次证监会的救市政策来看，新股扩容和资金扩容成了跷跷板的两端。政府不断通过调控这两个主题，调节着股价的涨与跌。

当股市好，国家就多发新股，当股市不好，国家多引资金入市。

最近的IPO恢复，可以明显看到新闻对这两方面的报道。首先恢复IPO，同时结束配资清理，并且降低融资担保率，鼓励融资买入股票。

不过2014年底的牛市，证监会可能希望让新股发行和资金入市同时增长，有对冲效果，形成慢牛。结果2015年年中，出了馊主意。

突然切断资金扩容这一端，结果新股还使劲发。本意是希望让股价涨速慢下来，但执行速度太快，行政干预太强，造成这艘船不平衡了，于是，船翻了。

现在重启IPO，重新鼓励融资，但愿这次开船开得慢一点。

最后
===================

从上面的数据来看，IPO和股价并没有直接的关联关系。

中国股市从诞生之日起，就是带着政治任务的，无论是当年的为国有企业解困，当中的为信托业解困，还是今年的为经济危机解困，或是未来的为万众创新提供资金支持，都以一种群众运动似的狂欢在一遍又一遍的进行着。

从历史来看，暂停IPO几乎从来没有改变熊市的趋向，但重启IPO之后，有时会人造一个牛市跟着。

此时谨记中央台的提醒：“注意安全行驶、低速行驶，与前车保持足够的距离”。
