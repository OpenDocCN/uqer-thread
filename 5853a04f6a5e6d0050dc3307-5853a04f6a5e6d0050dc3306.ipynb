### 1、打新简单介绍
---
- 个人投资者通过购买股票（不包括基金）就有了申购新股的资格，上交所1万起，深交所5千起
- 由于制度、市场环境等因素，一般情况下，新股上市后都会有3倍左右的收益
- 所以，买股票打新股成为当下非常时髦的一类投资策略

### 2、统计从2014年以来新股的年化收益情况
---
- 由于新股的发行是断断续续的，不同新股的中签率也不一样，所以一般统计某一时间段的新股年化收益率
- 新股年化收益率主要和三个因素相关：**中签率、发行价、开板收益率**
- `下面将分三步介绍计算过程，对细节不感兴趣的同学可以直接看第三部分`

### A、统计区间中签率
---
- 不同股票有不同中签率，统计某一段时间内的数据更有价值
- 下面的例子以20160101~20160701为区间段，同时也区分沪市和深市，来简单看一下市场的平均中签率

```python
import numpy as np        
import pandas as pd
import lib.Datayes as Datayes
from datetime import datetime as dt
from CAL.PyCAL import font

start_date = '20160101'
end_date = '20160701'
ndays = Datayes.delta_days(start_date, end_date)  
tickers = Datayes.get_IPOs(end_date, ndays)  # tickers为区间段内所有发行股票的ticker
sh_tickers, sz_tickers = Datayes.seperate_tickers(tickers)  # 将tickers按沪市和深市汇总

res = DataAPI.EquIPOGet(ticker=tickers, field='ticker,onlineIssueLottoRatio,issuePrice,listDate', pandas="1").set_index('ticker')
res['listDate'] = res['listDate'].apply(lambda x: x.replace('-', ''))
lotto_ratio = res['onlineIssueLottoRatio']
print '全市场平均中签率: ' + '%.3f%%' % (100*lotto_ratio.mean())
print '上交所平均中签率: ' + '%.3f%%' % (100*lotto_ratio[sh_tickers].mean())
print '深交所平均中签率: ' + '%.3f%%' % (100*lotto_ratio[sz_tickers].mean())
print '-----------------------------'
print '说明：上交所一个单位为1000股，深交所一个单位为500股！！！'
```

### B、计算新股上市后的开板收益率
---
- 做个一般的假设，以收盘价来进行买卖（毕竟也没那么多时间一直盯着新股）
- 所以开板收益率可以用新股上市后第一次收盘价没涨停来刻画（实际中你也可以每个交易日下午2：58去看看，若没有涨停就卖出）

```python
largest_ndays = 50  # 默认的最大连板天数
res['ipo_openDate'] = np.nan  # 新股开板日期 
res['ipo_openPrice'] = np.nan  # 新股开板收盘价
for t in res.index:
    starts = res.loc[t, 'listDate']
    ends = Datayes.shift_date(starts, largest_ndays, 'forward')
    data = DataAPI.MktEqudGet(ticker=t,beginDate=starts,endDate=ends,field='tradeDate,closePrice',pandas="1").sort('tradeDate').set_index('tradeDate')
    data.index = map(lambda x: x.replace('-',''), data.index)
    for d in data.index:
        try:  # 数据部门缺数据
            uplimit = DataAPI.MktLimitGet(ticker=t,tradeDate=d,field='limitUpPrice',pandas="1").loc[0, 'limitUpPrice']
        except:
            if d == data.index[0]:
                print t + ' 在 ' + d + '(上市首日) 缺乏limitUpPrice数据！！！'
            else:
                print t + ' 在 ' + d + '(非上市首日) 缺乏limitUpPrice数据！！！'
            continue
        if uplimit != data.loc[d, 'closePrice']:
            closePrice = data.loc[d, 'closePrice']
            break
        if d == data.index[-1]:  # 当循环到最后一天还没开板时
            if d < Datayes.shift_date(dt.today().strftime('%Y%m%d'), 0):
                print 'largest_ndays过小， ' + t + ' 没有找到开板价格！！！'
            else:
                closePrice = data.loc[d, 'closePrice']
    res.loc[t, 'ipo_openDate'] = d        
    res.loc[t, 'ipo_openPrice'] = closePrice
res['ipo_return'] = res['ipo_openPrice'] / res['issuePrice'] - 1
```

#### 说明几点
---
- 如上的具体过程可以不用太在乎，干的事情就是按照上述的假说统计出了一些计算新股收益率过程中的一些信息
- 如下所示，包括了每一个新股的`网上中签率、发行价、上市日、开板日期、开板价格以及开板收益率`
- 至此，便可以得到在20160101~20160701之间新股的中签率，整体开板收益率等数据，有点吓人，今年上半年新股平均收益率为429%！！！

### C、引入发行价，计算区间段内新股年化收益率
---
- 以1万元配售市值为例，计算区间段内沪深两市期望打新收益额
- 根据交易所摇号规则，一个号理论上只可能是中一签或者不中，且中签率普遍都低于0.001
- 个股期望收益额 = 中签率 X 中签收益额（中签收益额 = 发行价 X 中签股数 X 开板收益率）
- 对于10万配售市值来说，由于只可能中一签或者不中，所以个股期望收益额相比1万来说多了10倍，这也反映了打新年化收益率没有规模递减效益（在顶格申购范围内）
- `结果如市场预期，两市打新年华收益率大概在15%左右，但请记住这是今年上半年的区间，新股发行速度偏缓慢而且是慢熊行情，后面的结果会让你大吃一惊！！！`

```python
sh_res = res.loc[sh_tickers]
sz_res = res.loc[sz_tickers]
sh_res['profit'] = sh_res['onlineIssueLottoRatio'] * 1000 * sh_res['issuePrice'] * sh_res['ipo_return']
sz_res['profit'] = 2 * sz_res['onlineIssueLottoRatio'] * 500 * sz_res['issuePrice'] * sz_res['ipo_return']
# print '上交所区间内期望收益额: ' + '%.2f' % sh_res['profit'].sum()
print '上交所区间内年化收益率: ' + '%.2f%%' % (sh_res['profit'].sum() / Datayes.delta_days(start_date, end_date) * 250 / 100)
# print '深交所区间内期望收益额: ' + '%.2f' % sz_res['profit'].sum()
print '深交所区间内年化收益率: ' + '%.2f%%' % (sz_res['profit'].sum() / Datayes.delta_days(start_date, end_date) * 250 / 100)
```

### 3、统计打新的历史年化收益率
---
- 按照如上的方式，以每半年为一个区间段，统计从14年以来新股的年化收益率

```python
date_list = ['20140101', '20140701', '20150101', '20150701', '20160101', '20160701', '20161216']
data_lotto_ratio = pd.DataFrame(index=date_list[1:], columns=['全市场','上交所','深交所'], data=0.0)
data_ipo_return = pd.DataFrame(index=date_list[1:], columns=['全市场','上交所','深交所'], data=0.0)
data_annualize_return = pd.DataFrame(index=date_list[1:], columns=['全市场','上交所','深交所'], data=0.0)
data_ipo_numbers = pd.DataFrame(index=date_list[1:], columns=['上交所','深交所'], data=0.0)
for i in range(len(date_list)-1):
    starts = date_list[i]
    ends = date_list[i+1]
    try:
        tmp = Datayes.get_period_lotto_statistics(starts, ends, returns='all')
    except Exception, e:
        print e
        continue
    data_lotto_ratio.loc[ends] = tmp['statistics_data'].loc['平均中签率']
    data_ipo_return.loc[ends] = tmp['statistics_data'].loc['平均开板收益率']
    data_annualize_return.loc[ends] = tmp['statistics_data'].loc['年化收益率']
    data_ipo_numbers.loc[ends] = [tmp['sh_raw_data'].shape[0], tmp['sz_raw_data'].shape[0]]
    
data_annualize_return['两市发行新股数'] = data_ipo_numbers.sum(axis=1)
data_annualize_return['每只股票年化收益率'] = data_annualize_return['全市场'] / data_annualize_return['两市发行新股数']
```

#### 这里展示两个结果：区间打新年华收益率 + 区间平均中签率
---
- 如下，上面一个为中签率，下面一个为打新年华收益率
- 统计结果也区分沪市和深市
- **重大发现1**：`在14年的时候中签率原来这么高，最高时有1%！！！`现在人多了，只有万4了。。。
- **重大发现2**：`打新年华收益率在15年上半年时到过456%！！！`现在不是牛市了，新股发行数量也没之前多了，但仍然有28%，最低的时候也有14.3%！！！

### 5、打新风险点&敏感性分析
---
- IPO暂停：预期年化收益率和新股发行个数密切相关
        - 在0.0005中签率下，一年发行150只左右，年化20%，但按照最近几个月的节奏，年化40%都快了
- 中签率降低：随着打新热，中签率可能会急速下降，在现有的已启用回拨机制的情况下，中签率基本位于0.0005左右
        - 0.0005对应大概年化20%
        - 若提高到0.001对应年化40%
        - 若降低到0.00001对应年化4%，但这个在短期内概率不高
- 新股收益率降低：这个概率不大，为防止破发，政策明文规定新股发行价对应的PE不得高于行业平均 & PE绝对值不得高于23
- 股灾4.0：股灾会导致配售市值出现严重缩水，从目前目前情况来看，上证综指若再跌20%便至2500点，概率也比较低
