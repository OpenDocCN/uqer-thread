# 价格时滞

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import scipy.stats as st
import matplotlib
import matplotlib.pyplot as plt
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm
```

```python
universe = DataAPI.EquGet(equTypeCD='A')['secID']
start_date = '20110101'
end_date = '20171201'         #注意：起始与截至日期之间需隔开大于20天
back_day_num = 5              #时滞天数
storehouse_period = 20            #20日调仓
test_universe = universe#[:10]    #为节省时间，先取部分股票进行实验
```

```python
#优化效率（读取的效率）
#各股票的收益率表
#chgPct_df = pd.DataFrame()        
chgPct_df = DataAPI.MktEqudGet(secID=universe,beginDate=start_date,endDate=end_date,field=['tradeDate','secID',"chgPct"],pandas="1")#.set_index('tradeDate').rename(columns={"chgPct":stock})
chgPct_df = chgPct_df.pivot_table(index='tradeDate',columns='secID',values="chgPct").iloc[back_day_num:,:]#.reset_index().rename(columns={chgPct_df.columns[0]:'tradeDate'})

print 'aaaa'
#chgPct_df.to_csv('stocks_chgPct_df(20120101~20171201).csv')
print 'chgPct_df ok'
#print chgPct_df.head()

#A股市场日收益率表
market_chgpct_df = DataAPI.MktIdxdGet(indexID=u"000002.ZICN",beginDate=start_date,endDate=end_date,field=['tradeDate','CHGPct'],pandas="1").rename(columns={'CHGPct':'CHGPct0'})             
for i in range(1,back_day_num+1):
    market_chgpct_df.loc[1:,'CHGPct%d'%i] = market_chgpct_df['CHGPct%d'%(i-1)].values.tolist()[:-1]    
market_chgpct_df = market_chgpct_df.iloc[back_day_num:,:].reset_index(drop=True)           #A股市场日收益率及后推表
print 'market_return is ok'

'''
#获取股票滚动收益
REVS20_df = pd.DataFrame(index=chgPct_df.index)
for tradeDate in chgPct_df.index[:5]:
    signal_tradeDate_df = DataAPI.MktStockFactorsOneDayGet(secID=universe,tradeDate=tradeDate,field=u"secID,REVS20",pandas="1").rename(columns={'REVS20':tradeDate}).set_index('secID')
    REVS20_df = pd.concat([REVS20_df,signal_tradeDate_df],axis=0)
REVS20_df = REVS20_df.iloc[(storehouse_period-1)*2:]
print 'len(REVS20_df) is ok'
print 'hhh'
'''

time_late_price_df = pd.DataFrame(columns=chgPct_df.columns)   #价格时滞因子表
for i in range(len(market_chgpct_df)-storehouse_period+1):
    data_line = []
    #print 'jjd'
    for stock in chgPct_df.columns:
        chgPct = chgPct_df.iloc[i:i+storehouse_period][stock].values
        market_chgpct = market_chgpct_df.iloc[i:i+storehouse_period,-6:].values
        model = sm.OLS(chgPct,sm.add_constant(market_chgpct)).fit()
        time_late_price = 1-model.params[1]/sum(model.params[1:])
        
        data_line.append(time_late_price)

    time_late_price_df.loc[chgPct_df.index[i+storehouse_period-1],:] = data_line
time_late_price_df = time_late_price_df.iloc[:-(storehouse_period-1)]

#time_late_price_df.to_csv('time_late_price_df(20120101~20171201).csv')
print 'time_late_price_df is ok'
#print time_late_price_df.head()
#print REVS20_df.index,len(REVS20_df)
```

```python
#寻找与时滞因子相关的指标
for factor in ['EquityFixedAssetRatio','EquityToAsset','Volatility','NPParentCompanyGrowRate','FY12P','ROE',]:
    factor_df = pd.DataFrame(columns=time_late_price_df.columns)
    for tradeDate in time_late_price_df.index:
        factor_sig_stock_df = DataAPI.MktStockFactorsOneDayGet(tradeDate=tradeDate,secID=chgPct_df.columns,field=u"secID,"+factor,pandas="1").rename(columns={'secID':tradeDate}).T
        factor_df = pd.concat([factor_df,factor_sig_stock_df],axis=0)
        print factor_sig_stock_df
    aim_values_df = time_late_price_df
    signal_factor_analsis_df = pd.concat([pd.Series(factor_df.stack(dropna=False).tolist()),pd.Series(aim_values_df.stack().tolist())],axis=1)
    corr_value = signal_factor_analsis_df.iloc[:,0].corr(signal_factor_analsis_df.iloc[:,1])
    print '%s'%factor+'corr_value:',corr_value
```

```python
#优化因子
#单因素分析，观察因子与其他因子的关系，并将相关性强的找出来，进行中性化
#日换手率，流通市值，总市值   turnoverRate，negMarketValue，marketValue
i = 1 #第i个图
for factor in ['turnoverRate','negMarketValue','marketValue']:
    factor_df = pd.DataFrame(index=time_late_price_df.index)
    for stock in time_late_price_df.columns:
        factor_sig_stock_df = DataAPI.MktEqudAdjGet(secID=stock,beginDate=time_late_price_df.index[0],endDate=time_late_price_df.index[-1],field='tradeDate,'+factor,pandas="1").rename(columns={factor:stock}).set_index('tradeDate')
        factor_df = pd.concat([factor_df,factor_sig_stock_df],axis=1)
    #print factor_df
    aim_values_df = time_late_price_df             #价格时滞因子表
    signal_factor_analsis_df = pd.concat([pd.Series(aim_values_df.stack(dropna=False).tolist()),pd.Series(factor_df.stack(dropna=False).tolist())],axis=1).rename(columns={0:'price_delay',1:factor}).dropna()
    model = ols('price_delay ~ '+factor,signal_factor_analsis_df).fit()
    anovat = anova_lm(model)
    print anovat
    corr_value = signal_factor_analsis_df.iloc[:,0].corr(signal_factor_analsis_df.iloc[:,1])
    print 'corr_value:',corr_value
    #plt.figuresize(1)
    plt.subplot(2,2,i)
    plt.scatter(signal_factor_analsis_df.iloc[:,0],signal_factor_analsis_df.iloc[:,1])
    i += 1
```
