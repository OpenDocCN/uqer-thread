## 高转送股票的基本特征：
##### 1、 每股资本公积高于市场均值
##### 2、 每股未分配利润高于市场均指
##### 3、 上市时间短
##### 4、 总股本小
##### 5、 股价相对较高

##### 研报中还提及了高转送股票的成长特征（每股净资产、每股收益、每股现金流），本文没有考虑，有待后续补充。

##### 根据高送转股票的基本特征，建立如下模型：
##### 第一步：每年 11 月 1 日,选取创业板股票作为样本空间
##### 第二步：对样本空间内 A 股股票按照每股资本公积、每股未分配利润、 股价、总股本从高到低进
##### 行排序,分成 10 组。最高组的股票在对应指标项打 10 分,次高组打 9 分,依次至最低组打 1 分。
##### 第三步： 对样本空间内 A 股股票按照上市时间长短,进行打分。2015年上市，10分；2014年上市，9分；2013年上市，8分...
##### 第四步： 五项总分合计排序，取出前50的股票作为买入列表。

```python
import pandas as pd
import math
import numpy as np
import datetime as dt
from pandas import Series, DataFrame, isnull
from datetime import timedelta
from CAL.PyCAL import *
cal = Calendar('China.SSE')

def signal_gzs_calc(universe, current_date):
    
    # current_date = dt.datetime(int(current_date[0:4]), int(current_date[5:7]), int(current_date[8:10]))   
    today = Date.fromDateTime(current_date)
    # today = current_date
    start_date = (today - Period('1Y')).toDateTime().strftime('%Y%m%d')
    end_date = today.toDateTime().strftime('%Y%m%d')
    
    data1=DataAPI.FdmtBSAllLatestGet(secID=universe,reportType='Q3',beginDate=start_date,endDate=end_date,field=['actPubtime','secID','capitalReser','retainedEarnings'],pandas="1")                                  #获取时间节点内的三季报数据，数据来自合并资产负债表

    data2=DataAPI.EquShareGet(secID=universe,beginDate=start_date,endDate=end_date,partyID=u"",field=['secID','changeDate','totalShares','floatA'],pandas="1")# 获取当前节点的无限售流通股数量，此处可换用nonrestfloatA：无限售流通A股，结论会显著不同。数据来自公司股本变动
    data2 = data2.sort(columns=['secID', 'changeDate'], ascending=[True, True])   # 排序
    data2 = data2.drop_duplicates(subset='secID', take_last=True)            # 按照股票代码去掉重复项
    
    data4 = DataAPI.EquGet(secID=universe,equTypeCD=u"A",listStatusCD=u"",field=['secID','secShortName','listDate'],pandas="1")
    data4['Score4'] = data4['listDate'].apply(lambda x: int(x[0:4])-2005)      # 
    
    yesterday = cal.advanceDate(today,'-1B',BizDayConvention.Unadjusted).strftime('%Y%m%d')
    data5 = DataAPI.MktEqudAdjGet(secID=universe,tradeDate=yesterday,field=['secID','closePrice'],pandas="1")
    data5 = data5.sort(columns=['secID', 'closePrice'], ascending=[True, True])   # 排序

    data=pd.merge(data1,data2)
    data['percapitalReser']=data['capitalReser']/data['floatA']           # 每股资本公积
    data['perretainedEarnings']=data['retainedEarnings']/data['floatA']      # 每股未分配利润

    h=len(data)                                    # 当前CYB共481只股票
    list2=[(10-10*x/h) for x in range(0,h)] 
    data=data.sort(columns='percapitalReser',ascending=0)        # 为每股资本公积进行降序重排
    data['Score1']=pd.Series(list2,index=data.index)           # 并赋值10-1


    data=data.sort(columns='perretainedEarnings',ascending=0)    # 为每股未分配利润进行降序重排赋值 
    data['Score2']=pd.Series(list2,index=data.index)        

    data=data.sort(columns='totalShares',ascending=1)         #为总股本数进行升序重排赋值
    data['Score3']=pd.Series(list2,index=data.index) 
    
    data4['Score4'] = data4['listDate'].apply(lambda x: int(x[0:4])-2005)
    
    data5=data5.sort(columns='closePrice',ascending=0)
    data5['Score5'] = pd.Series(list2,index=data.index)
    
   
    data['Score']=data['Score1']+data['Score2']+data['Score3']+data4['Score4']+data5['Score5']


    data=data.sort(columns='Score',ascending=0)

    pb_signal = data[['secID', 'Score']]
    score=pb_signal.head(50)
    return list(score['secID'])
```
