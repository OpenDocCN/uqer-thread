# 因子数据调取与处理

# Fama-Franch 三因子模型之规模因子

规模因子大盘收益减去小盘收益而得出的因子，简单计算的话可以使用中证1000的平均收益减去上证50， 然而这样的计算我认为有失公允。单纯从绝对规模来衡量大小盘，基本肯定会叠加value与growth的特征与行业的特征。A股的大盘股中金融占了绝大多数，上证50与中证1000的比较相当于是金融与中小企业的比较，有失公允。

由此，想了一个方法，从分行业来提取各行业中大盘与小盘，通过平均大小盘的差异来制造规模因子。 取了Wind行业指数，取成分股，剔除B股，按年交易量平均排序，取前10%为小盘，取后10%为大盘。再取对应股票收益进行平均。最后对所有行业进行平均得到规模因子。

```python
import pandas as pd
import datetime as dt
import numpy as np
```

## 1. 取Wind行业成分股列表

```python
sheetNames = pd.ExcelFile('./年度报告/分行业股票列表.xlsx').sheet_names
```

```python
def aveTradeVolumn(sheet_name):
    """
    取成分股的年均成交量，返回DataFrame
    """
    df = pd.DataFrame()
    industry = pd.read_excel('./年度报告/分行业股票列表.xlsx', sheetname = sheet_name)
    for tickerCode in industry[industry[u'代码'].apply(lambda x: (x[0] != u'2')&(x[0] != u'9'))][u'代码']:
        df[tickerCode] = DataAPI.MktEqudGet(secID=u"",
                           ticker = tickerCode[:-3],
                           tradeDate=u"",
                           beginDate=u"20171231",endDate=u"20181231",
                           isOpen="",field=u"",pandas="1")['dealAmount']
    df.index = DataAPI.MktEqudGet(secID=u"",
                           ticker = tickerCode[:-3],
                           tradeDate=u"",
                           beginDate=u"20171231",endDate=u"20181231",
                           isOpen="",field=u"",pandas="1")['tradeDate']
    return df.mean(axis=0)

def SmallAndBig(sheet_name, n):
    """
    按排序的百分数取大小盘股票列表list
    """
    series = aveTradeVolumn(sheet_name).sort_values()
    N = int(np.ceil(series.size*n))
    return series[:N].index.tolist(), series[-N:].index.tolist()

def get_small_big(sheet_name,n=0.1):
    """
    根据大小盘list中的股票取收益，返回两个dataframe，小盘与大盘
    """
    small,big = SmallAndBig(sheet_name,n)
    small_df = pd.DataFrame(index=pd.date_range(start='2014-12-31', end='2018-12-31',freq='B'))
    big_df = pd.DataFrame(index=pd.date_range(start='2014-12-31', end='2018-12-31',freq='B'))
    for s, b in zip(small, big):
        try:
            small_df = small_df.join(DataAPI.MktEqudGet(secID=u"",
                                   ticker = s[:-3],
                                   tradeDate=u"",
                                   beginDate=u"20141231",endDate=u"20181231",
                                   isOpen="",field=u"",pandas="1").set_index('tradeDate')['closePrice'].rename(s))
            big_df[b] = big_df.join(DataAPI.MktEqudGet(secID=u"",
                                   ticker = b[:-3],
                                   tradeDate=u"",
                                   beginDate=u"20141231",endDate=u"20181231",
                                   isOpen="",field=u"",pandas="1").set_index('tradeDate')['closePrice'].rename(b))
        except:
            pass
    return small_df, big_df

def smb_returns(sheet_name,n=0.1):
    """
    小盘股收益 - 大盘股收益，得到行业中的规模因子
    
    """
    small_df, big_df = get_small_big(sheet_name,n=0.1)
    small_returns = small_df.fillna(method='bfill').fillna(method='ffill').pct_change().mean(axis=1)
    big_returns = big_df.fillna(method='bfill').fillna(method='ffill').pct_change().mean(axis=1)
    SMB_returns = small_returns - big_returns

    return SMB_returns
```

# 规模因子走势图

```python
# 提取各行业规模因子组成DataFrame

df = pd.DataFrame()
n=0
for sheet_name in sheetNames:
    try:
        df[sheet_name] = smb_returns(sheet_name,0.1)
        n+=1
        print(n)
    except:
        pass
```

## 分行业规模因子走势

```python
df_nav = (df+1).fillna(1).cumprod()
g = df_nav.plot(figsize=(15,5))
g.legend(range(df_nav.shape[1]));
```

## 平均规模因子走势

```python
(df.mean(axis=1)+1).fillna(1).cumprod().plot(figsize=(15,5));
```
