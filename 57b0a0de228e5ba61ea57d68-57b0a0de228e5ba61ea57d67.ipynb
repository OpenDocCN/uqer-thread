# 

上周发布的基于Fama-French三因子和五因子模型的多因子策略，在一定程度上说明了基于现代金融学资产定价理论的基本面量化策略在中国股市的无效性。因此，本周发布的量化策略考虑风格轮动策略，即通过一系列的宏观与市场指标，判别小盘股、大盘股、价值股与成长股在
不同市场条件下的相对表现。

首先，将沪市所有A股（共计1100只）每月按总市值和市净率分为四组：group 0-小盘股（低市值）&价值股（低PB），group 1-小盘股（低市值）&成长股（高PB），group 2-大盘股（高市值）&价值股（低PB），大盘股（高市值）&成长股（高PB）。其中总市值前20%为高市值，后20%为低市值股票，市净率前20%为高PB，后20%为低PB。

本策略选用宏观指标为：PPI、CPI、净出口额、货币供应量（M1/M2)、外商直接投资、房地产开发综合景气指数、汇率、利率（银行隔夜拆借利率）和黄金价格。
选用市场指标为：大小盘风格收益（上证50与中证500收益率之差）、三个月动量、波动率。
通过当月的月度指标，运用Logistic回归及SVM预测下月收益率最高的组，样本期为74个月，以滚动窗口进行预测，交易期为2013年7月至2016年6月共36个月（三年）。

最终预测结果显示，Logistic的样本外预测表现优于SVM，其中样本内预测准确率62%，样本外预测准确率为58%。

根据预测结果，选取下月预期收益最高的组内的所有股票，以等权方式构建单一做多投资组合。 回测区间为2013年7月1日至2016年6月30日，策略参考标准为上证综指，证券池为沪市A股，调仓周期为一月，调仓日期为月初。回测结果显示，本策略的年化收益远高于上证综指，约为其三倍左右。

```python
from CAL.PyCAL import *
import pandas as pd
import numpy as np
from sklearn import preprocessing
from sklearn import cross_validation
from sklearn import svm
from sklearn import linear_model
from scipy import stats
import itertools         

# 获取宏观指标对应的宏观数据
PPI=DataAPI.ChinaDataPPIGet(indicID=u"M030000147",beginDate='20070401',endDate='20160531',field=['periodDate','dataValue'],pandas="1")
CPI=DataAPI.ChinaDataCPIGet(indicID=u"M030000012",beginDate='20070401',endDate='20160531',field='dataValue',pandas="1")
NX=DataAPI.ChinaDataForeignTradeGet(indicID=u"M090000003",beginDate='20070401',endDate='20160531',field='dataValue',pandas="1")
M1=DataAPI.ChinaDataMoneyStatisticsGet(indicID=u"M100000004",beginDate='20070401',endDate='20160531',field='dataValue',pandas="1")
M2=DataAPI.ChinaDataMoneyStatisticsGet(indicID=u"M100000005",beginDate='20070401',endDate='20160531',field='dataValue',pandas="1")
M1_M2=M1/M2
FI=DataAPI.ChinaDataFDIGet(indicID=u"M090000030",beginDate='20070401',endDate='20160531',field='dataValue',pandas="1")
macro0=pd.concat([PPI,CPI,NX,M1_M2,FI],axis=1)
macro0["periodDate"]=macro0["periodDate"].apply(Date.parseISO)
macro0['yearmonth']=macro0['periodDate'].apply(Date.month)+macro0['periodDate'].apply(Date.year)*100
del macro0['periodDate']
Property=DataAPI.ChinaDataRealEstateGet(indicID=u"M080000003",beginDate='20070401',endDate='20160531',field=['periodDate','dataValue'],pandas="1")
Property["periodDate"]=Property["periodDate"].apply(Date.parseISO)
Property['yearmonth']=Property['periodDate'].apply(Date.month)+Property['periodDate'].apply(Date.year)*100
del Property['periodDate']
macro1=pd.merge(macro0,Property,on='yearmonth',how='outer')

ER=DataAPI.ChinaDataExchangeRateGet(indicID=u"M110000003",beginDate='20070401',endDate='20160531',field=['periodDate','dataValue'],pandas="1")
ER["periodDate"]=ER["periodDate"].apply(Date.parseISO)
ER['yearmonth']=ER['periodDate'].apply(Date.month)+ER['periodDate'].apply(Date.year)*100
ER=ER.groupby('yearmonth').head(1)
del ER['periodDate'] 
IR=DataAPI.ChinaDataInterestRateSHIBORGet(indicID=u"M120000064",beginDate='20070401',endDate='20160531',field=['periodDate','dataValue'],pandas="1")
IR["periodDate"]=IR["periodDate"].apply(Date.parseISO)
IR['yearmonth']=IR['periodDate'].apply(Date.month)+IR['periodDate'].apply(Date.year)*100
IR=IR.groupby('yearmonth').head(1)
del IR['periodDate']    
Gold=DataAPI.ChinaDataGoldClosePriceGet(indicID=u"M140000090",beginDate='20070401',endDate='20160531',field=['periodDate','dataValue'],pandas="1")
Gold["periodDate"]=Gold["periodDate"].apply(Date.parseISO)
Gold['yearmonth']=Gold['periodDate'].apply(Date.month)+Gold['periodDate'].apply(Date.year)*100
Gold=Gold.groupby('yearmonth').head(1)
del Gold['periodDate']  
macro2=pd.merge(ER,IR,on='yearmonth')
macro3=pd.merge(macro2,Gold,on='yearmonth')
macro=pd.merge(macro3,macro1,on='yearmonth')
macro.set_index('yearmonth',inplace=True)
macro.sort_index(inplace=True)
macro.fillna(method='bfill',inplace='True')
```

```python
# 获取市场指标对应的市场数据
def month_ret(df):
    df['yearmonth']=df['tradeDate'].apply(Date.month)+df['tradeDate'].apply(Date.year)*100
    df_ret1=df.groupby('yearmonth').tail(1)
    df_ret2=pd.DataFrame()
    df_ret2=df_ret1['yearmonth']
    df_ret2['month_ret']=df_ret1['closeIndex'].pct_change()
    df_ret2=df_ret2.dropna(axis=0,how='any') 
    return df_ret2['month_ret']

prc_500=DataAPI.MktIdxdGet(indexID="000905.ZICN",beginDate="20070101",endDate="20160531",field=['secShortName',"tradeDate","closeIndex"],pandas="1")
prc_500["tradeDate"]=prc_500["tradeDate"].apply(Date.parseISO)
prc_500.index=prc_500["tradeDate"]
ret_500=month_ret(prc_500)#中证500收益率

prc_50=DataAPI.MktIdxdGet(indexID="000016.ZICN",beginDate="20070101",endDate="20160531",field=['secShortName',"tradeDate","closeIndex"],pandas="1")
prc_50["tradeDate"]=prc_50["tradeDate"].apply(Date.parseISO)
prc_50.index=prc_50["tradeDate"]
ret_50=month_ret(prc_50)#上证50收益率

ret_style=ret_50-ret_500#风格收益率

prc_all=DataAPI.MktIdxdGet(indexID="000001.ZICN",beginDate="20070101",endDate="20160531",field=['secShortName',"tradeDate","closeIndex"],pandas="1")
prc_all["tradeDate"]=prc_all["tradeDate"].apply(Date.parseISO)
prc_all.index=prc_all["tradeDate"]
momentum=pd.rolling_mean(month_ret(prc_all), window=3)#三个月动量

prc_all['ret_day']=prc_all['closeIndex'].pct_change()
prc_all['yearmonth']=prc_all['tradeDate'].apply(Date.month)+prc_all['tradeDate'].apply(Date.year)*100
volatility=prc_all.groupby('yearmonth')['ret_day'].std()#波动率

df_temp=pd.concat([ret_style,momentum],axis=1)
dates=pd.DataFrame()
dates['dates']=df_temp.index
df_temp.index=dates['dates'].apply(Date.month)+dates['dates'].apply(Date.year)*100
df_indicator=pd.concat([macro,df_temp,volatility],axis=1)
df_indicator.columns=['ER','IR','Gold','PPI','CPI','NX','M1/M2','FI','Property','ret_style','momentum','volatility']
df_indicator=df_indicator.dropna(axis=0,how='any')
```

```python
# 获取沪市A股行情数据
universe = set_universe(IdxCN.Idx000001) 
df_ret_temp=DataAPI.MktEqudAdjGet(secID=universe,beginDate="20070401",endDate="20100630",field=["secID","tradeDate","closePrice","marketValue"],pandas="1")
df_ret_temp=df_ret_temp.append(DataAPI.MktEqudAdjGet(secID=universe,beginDate="20100701",endDate="20130630",field=["secID","tradeDate","closePrice","marketValue"],pandas="1"))
df_ret_temp=df_ret_temp.append(DataAPI.MktEqudAdjGet(secID=universe,beginDate="20130701",endDate="20160630",field=["secID","tradeDate","closePrice","marketValue"],pandas="1"))
df_ret_temp['yearmonth']=[int(x[:4])*100+int(x[5:7]) for x in df_ret_temp['tradeDate']]
df_ret_temp['month']=[int(x[5:7]) for x in df_ret_temp['tradeDate']]
df_ret_temp['year']=[int(x[:4]) for x in df_ret_temp['tradeDate']]
df_ret_temp['accMonth']=np.nan
df_ret_temp['accMonth'][df_ret_temp['month']<=3]=12+(df_ret_temp['year']-1)*100
df_ret_temp['accMonth'][(df_ret_temp['month']<=6) & (df_ret_temp['month']>3)]=3+df_ret_temp['year']*100
df_ret_temp['accMonth'][(df_ret_temp['month']<=9) & (df_ret_temp['month']>6)]=6+df_ret_temp['year']*100
df_ret_temp['accMonth'][(df_ret_temp['month']<=12) & (df_ret_temp['month']>9)]=9+df_ret_temp['year']*100
```

```python
# 获取上市公司的净资产数据
TshEquity=DataAPI.FdmtBSAllLatestGet(secID=universe,endDate="20160630",beginDate="20070301",field=['secID','endDate','TShEquity'],pandas="1")
TshEquity['accMonth']=[int(x[:4])*100+int(x[5:7]) for x in TshEquity['endDate']]
```

```python
#计算得到各股票的市净率
df_group=pd.merge(df_ret_temp,TshEquity,on=['secID','accMonth'])
df_group['PB']=df_group['marketValue']/df_group['TShEquity']
del df_group['accMonth']
del df_group['endDate']
del df_group['year']
del df_group['month']
del df_group['TShEquity']
df_group=df_group.dropna(how='any')

del df_ret_temp
del TshEquity
```

```python
#计算得到各股票的月收益率
df_group=df_group.groupby(['yearmonth','secID']).tail(1)
df_group['ret_month']=df_group.groupby('secID')['closePrice'].pct_change()
df_group=df_group.dropna(how='any')
```

```python
# 对股票按当月的总市值和市净率进行分组
df_group[['marketValue_rank','PB_rank']]=df_group.groupby('yearmonth')[['marketValue','PB']].rank(pct=True)
df_group['group']=np.nan
df_group['group'][(df_group['marketValue_rank']<0.2) & (df_group['PB_rank']<0.2)]=0
df_group['group'][(df_group['marketValue_rank']<0.2) & (df_group['PB_rank']>0.8)]=1
df_group['group'][(df_group['marketValue_rank']>0.8) & (df_group['PB_rank']<0.2)]=2
df_group['group'][(df_group['marketValue_rank']>0.8) & (df_group['marketValue_rank']>0.8)]=3
df_group=df_group.dropna(how='any')
group_ret=df_group.groupby(['yearmonth','group'],as_index=False)['ret_month'].mean()
```

```python
# 严格意义上对股票分组应当按照上月的总市值和市净率进行分组，但由于分组结果与按当月的总市值和市净率的分组结果基本相同，本策略选用当月的总市值和市净率进行分组。此处为按上月的总市值和市净率进行分组的相关代码。
# df_group['group_pre']=df_group['group'].shift(1)
# df_group1=df_group.drop(df_group.groupby('secID').head(1).index)
# group_ret_next=df_group1.groupby(['yearmonth','group_pre'],as_index=False)['ret_month'].mean()
```

```python
#得到下月收益最高的组
max_ret=group_ret.groupby('yearmonth',as_index=False)['ret_month'].max()
group_ret=pd.merge(group_ret,max_ret,on='yearmonth')
group_ret.columns=['yearmonth','group','ret_month','ret_max']
group_ret['isMax']=(group_ret['ret_month']==group_ret['ret_max'])
group_ret=group_ret.set_index('yearmonth')
result=pd.DataFrame()
result['group_max']=group_ret['group'][group_ret['isMax']==True]
```

```python
# 构造对解释变量进行标准化，并除去离群点的函数
def winsorize(data):
    data[data>np.percentile(data[0:window],99)]=np.percentile(data[0:window],99)
    data[data<np.percentile(data[0:window],1)]=np.percentile(data[0:window],1)
    mean=np.median(data[0:window])
    std=np.nanstd(data[0:window])
    return (data-mean)/std
```

```python
#通过Logistic回归与SVM，以滚动窗口进行预测。结果表明Logistic回归表现优于SVM，故此处选用Logistic回归。 
window=74
roll_pred=[]

for i in range(0,110-window):    
    x=df_indicator.iloc[i:i+window+1]
    z=pd.DataFrame()
    z['ret_style']=winsorize(x['ret_style'].values)#小盘股与大盘股收益率差
    z['momentum']=winsorize(x['momentum'].values)
    z['volatility']=winsorize(x['momentum'].values)
    z['property']=winsorize(x['Property'].values)
    z['M1/M2']=winsorize(x['M1/M2'].values)
    z['IR']=winsorize(x['IR'].values/12)
    z['ER']=winsorize(x['ER'].values)
    z['NX']=winsorize(x['NX'].values)
    z['PPI']=winsorize(x['PPI'].values)
    z['gold']=winsorize(x['Gold'].values)
    z['CPI']=winsorize(x['CPI'].values)
    z['FI']=winsorize(x['FI'].values)
    y = result[i:i+window]
    x1 = z[0:window]
    mdl = linear_model.LogisticRegression()
    # mdl=svm.SVC()
    mdl_fit = mdl.fit(x1,y)
    if i==0:
        pred_insample=mdl_fit.predict(x1)
    roll_pred.append(mdl_fit.predict(z.iloc[-1])[0])
```

```python
#得到样本内与样本外的预测准确率，并得到交易期每个月的股票种类。
pred_all=pd.concat((pd.DataFrame(pred_insample),pd.DataFrame(roll_pred)))
pred_all.columns=['group_pred']
pred_all['yearmonth']=result.index
df_stk=pd.merge(df_group,pred_all,on='yearmonth')
ratio_insample=float(np.sum((pred_all.values-result.values)[:window]==0))/74
print("样本内预测准确率：%0.2f" % ratio_insample)
ratio_out=float(np.sum((pred_all.values-result.values)[window:110]==0))/36
print("样本外预测准确率：%0.2f" % ratio_out)
```
