```python
#首先导入程序必备的库
from CAL.PyCAL import *
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
import matplotlib.pyplot as plt
from datetime import datetime

#设定起始日与结束日（默认为Today）
start_date = Date(2010,1,1)
end_date = Date.todaysDate()
#将起始日与结束日的格式转化成方便DataAPI的字符串形式
start_date = start_date.toDateTime().strftime('%Y%m%d')
end_date = end_date.toDateTime().strftime('%Y%m%d')

#选定某只股票，以“中国银行”为例
symbol = '601988.XSHG' 

#调用MktEqudGet数据接口，获取设定时间段以来的收盘价与前收盘价
data = DataAPI.MktEqudGet(secID=symbol,ticker=u"",beginDate=start_date,endDate=end_date,field=['secID','tradeDate','closePrice','preClosePrice'],pandas="1")

#计算日收益率，并添加一列到DataFrame
data['daily_return'] = data['closePrice']/data['preClosePrice'] - 1 

#设定每年的默认交易日数量（250天），并在DataFrame中添加“期望收益率”与“日度波动率”两列
window = 250
data['expected_return'] = pd.rolling_mean(data['daily_return'],window)
data['volatility'] = pd.rolling_std(data['daily_return'],window)
data['sharpe_ratio'] = data['expected_return'] / data['volatility'] * np.sqrt(window)

tmp = data[['tradeDate','sharpe_ratio']]
tmp = tmp.dropna()
#tmp = pd.Series(tmp['sharpe_ratio'].tolist(),index=tmp['tradeDate'])
#print tmp

#将Rolling_Window的年化夏普率作图
plt.plot([datetime.strptime(t,"%Y-%m-%d").date() for t in tmp['tradeDate']],tmp['sharpe_ratio'])
plt.ylabel('Rolling Annualized Sharpe Ratio')
```
