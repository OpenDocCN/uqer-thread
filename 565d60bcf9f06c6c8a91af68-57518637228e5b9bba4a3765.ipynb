```python
import numpy as np
import pandas as pd
import datetime as dt
from math import sqrt

dataG=DataAPI.MktEqudGet(secID=u"",ticker=u"601398",tradeDate=u"",beginDate=u"20131009",endDate=u"20151012",field=u"",pandas="1")
dataJ=DataAPI.MktEqudGet(secID=u"",ticker=u"601939",tradeDate=u"",beginDate=u"20131009",endDate=u"20151012",field=u"",pandas="1")
def multipl(a,b):
    sumofab=0.0
    for i in range(len(a)):
        temp=a[i]*b[i]
        sumofab+=temp
    return sumofab
def corrcoef(x,y):
    n=len(x)
    #求和
    sum1=sum(x)
    sum2=sum(y)
    #求乘积之和
    sumofxy=multipl(x,y)
    #求平方和
    sumofx2 = sum([pow(i,2) for i in x])
    sumofy2 = sum([pow(j,2) for j in y])
    num=sumofxy-(float(sum1)*float(sum2)/n)
    #计算皮尔逊相关系数
    den=sqrt((sumofx2-float(sum1**2)/n)*(sumofy2-float(sum2**2)/n))
    return num/den
data_closeG=dataG.closePrice
data_closeJ=dataJ.closePrice
print corrcoef(data_closeG,data_closeJ) 
```
