```python
# 利用优矿CAL分析库进行期权定价
# 选择二叉树定价引擎
# 这里有两条假设：（1）白糖期货10月28日涨幅为15.78%，没有涨停板；（2）定价时10月8日隐含波动率还是9月28日的波动率
from CAL.PyCAL import *
from CAL import *

def BinomialPrice(strike, volatility, optionType):
    # 设定过程为BSM
    process = BlackScholesMertonProcessConstant(underlying, 
                                                riskFree, 
                                                dividend, 
                                                volatility, 
                                                'Actual/Actual (ISMA)') 
    # 设定行权方式
    exercise = AmericanExercise(evaDate, exDate)
    payoff = PlainVanillaPayoff(optionType, strike)
    option = VanillaOption(payoff, exercise)
    # 定价引擎为二叉树
    engine = BinomialVanillaEngine(process, "crr", 400)
    option.setPricingEngine(engine)

    return option.NPV()    
```

```python
evaDate = Date(2018,10,8)
# 设置估值日期
SetEvaluationDate(evaDate)
# 设置到期日
exDate = Date(2018,11,26)
#找到9月28日标的价格
varPrice = DataAPI.MktFutdGet(secID=u"",ticker=u"SR901",tradeDate=u"20180928",beginDate=u"",endDate=u"",field=['closePrice'],pandas="1").values[0][0]
underlying = varPrice*(1+0.1578)#假定10月8日标的上涨15.78%
riskFree = 0.02
dividend = 0.00
data['closePrice1008'] = None

for i in range(len(data)):
    #print i
    if data.loc[i, 'class'] == 'SR901C':        
        optionType = Option.Call
    if data.loc[i, 'class'] == 'SR901P':        
        optionType = Option.Put    
    volatility = data.loc[i, 'impVol']
    strike = data.loc[i, 'strikePrice']
    try:
        data.loc[i, 'closePrice1008'] = BinomialPrice(strike, volatility, optionType)
    except:
        pass
    continue
```

```python
# 计算国庆期间每个合约的预估收益率
data['return'] = (data['closePrice1008'] / data['settlPrice'] -1) * 100
```
