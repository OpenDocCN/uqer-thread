# 僵尸股

```python
import pandas as pd
import numpy as np
from CAL.PyCAL import *
from pandas import DataFrame,Series
from datetime import datetime, timedelta

cal = Calendar('China.SSE')
end = (datetime.today() - timedelta(days=1)).strftime('%Y%m%d')  # 昨天
lastyear = cal.advanceDate(end, '-1Y', BizDayConvention.Following).strftime('%Y%m%d')
twoyears=cal.advanceDate(end, '-2Y', BizDayConvention.Following).strftime('%Y%m%d')
threeyears=cal.advanceDate(end, '-3Y', BizDayConvention.Following).strftime('%Y%m%d')

universe=set_universe('A')

date=DataFrame()
for secId in universe:
    try:
        lastyearEPS=DataAPI.MktStockFactorsDateRangeGet(secID=secId,ticker=u"",beginDate=lastyear,endDate=end,field=[u"secID",u"ticker",u"EPS"],pandas="1")
        lastTwoyearEPS=DataAPI.MktStockFactorsDateRangeGet(secID=secId,ticker=u"",beginDate=twoyears,endDate=lastyear,field=[u"secID",u"ticker",u"EPS"],pandas="1")
        lastThreeyearEPS=DataAPI.MktStockFactorsDateRangeGet(secID=secId,ticker=u"",beginDate=threeyears,endDate=twoyears,field=[u"secID",u"ticker",u"EPS"],pandas="1")
    
        #if(lastyearEPS["EPS"].mean()<0 and lastTwoyearEPS["EPS"].mean()<0 and lastThreeyearEPS["EPS"].mean()<0):
        if(lastyearEPS["EPS"].mean()<0 and lastTwoyearEPS["EPS"].mean()<0):
            
            row=pd.DataFrame([dict(secID=secId,ticker=lastyearEPS["ticker"][1],oneyear=lastyearEPS["EPS"].mean(),twoyear=lastTwoyearEPS["EPS"].mean(),threeyear=lastThreeyearEPS["EPS"].mean())])
            date=date.append(row,ignore_index=True)
    except:
        pass
date.to_csv("zobbie.csv")
```
