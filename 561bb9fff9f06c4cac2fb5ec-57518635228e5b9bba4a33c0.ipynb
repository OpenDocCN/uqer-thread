# 简单策略问题——国科大量化投资课小组作业1

本节课，魏老师简单介绍了一下量化投资的基本概念，并给我们提供了一个简单策略进行股票投资。
该策略简言之，利用五日均线与十日均线，判定是否买入卖出。
基于此，本小组暂定核心逻辑为“当前五日收盘均值大于前十日收盘价，当日需买入；反之则卖出；除此之外，不进行交易”。
由以上逻辑，我们使用python编程进行初步尝试。

根据程序结构，我们分为以下3部分。

数据导入及重整

在该部分中，我们提取编号为600030的中信证券作为考察股，导入从2013年10月1日至今的全部交易信息。基于10月1-7日为公休，8日才正式开盘，所以此时起始日期实际为10月8日。
在本策略中，我们仅关心收盘价及开盘价，所以仅需取交易信息中的两种价格信息。

```python
data = DataAPI.MktEqudGet(ticker="600030",beginDate="20131008")     # 导入600030的股票信息
data1 = data["closePrice"]                                          # 导出需要的列信息，此处为收盘价
data2 = data["openPrice"]
closePrice = data1.values                                           # 将列表信息转化为数组，这样便于之后的计算
openPrice = data2.values
```

判断买卖情况

在该部分，我们对比从10月8日开盘后第十一天至今，每天的前5日与前10日的平均值，判断明日开盘是否买卖。

```python
n = len(closePrice)
#print n
judge = []                                                                                              # 初始一个判断数组
for i in range(10,n+1):
    adv5 = (closePrice[i-1]+closePrice[i-2]+closePrice[i-3]+closePrice[i-4]+closePrice[i-5])/5          # 得出前5日的平均值
    adv10 = (closePrice[i-1]+closePrice[i-2]+closePrice[i-3]+closePrice[i-4]+closePrice[i-5]+closePrice[i-6]+closePrice[i-7]+closePrice[i-8]+closePrice[i-9]+closePrice[i-10])/10
    if adv5 >= adv10:
        judge.append(True)                                                                              # 当前5日的平均值大于前10日，在判断数组中输出一个True
    else:
        judge.append(False)
```

计算收益

在该部分，我们利用之前得到的判断买卖的数组，计算具体每一笔收益。
起始资产是1亿元，全部用来购买证劵，以后无论亏损，得到的所有资产均用来购买该股票。
触发交易的情况有两种：买，当日判断为True，昨日为False；卖，当日判断为False,昨日为True。均以当日开盘价进行买卖。

```python
import matplotlib.pyplot as plt
m = len(judge)
price = 100000000.0                                                 # 起始金额数
amount = 0.0                                                        # 股数
price0 = 0.0                                                        # 单次收益
price1 = 0.0                                                        # 买入价
price2 = 0.0                                                        # 卖出价
count = 0                                                           # 交易次数
Price = []
Diff = []
for j in range(1,m+1):
    if j == m:
        price0 = 0
    else:
        if (judge[j-1] == False and judge[j] == True):              # 当日可买
            price1 = openPrice[j+9]
            amount = price // price1
            count = count + 1
        elif (judge[j-1] == True and judge[j] == False):            # 当日可卖
            price2 = openPrice[j+9]
            price0 = (price2 - price1) * amount
            count = count + 1
    price = price + price0
    Price.append(0)
    Diff.append(price-100000000)

print("Profit and Loss: %f"%(price - 100000000))                    # 打印出最终盈亏值
print("Number of transactions : %d"%count)                          # 总交易次数

plt.plot(Price,color="Black",label="initial")                       # 零均线
plt.plot(Diff,color="green",label="Profit and Loss")                # 盈亏曲线
plt.xlabel("Date")
plt.ylabel("Price")

plt.show()
```
