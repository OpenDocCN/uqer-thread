# Python金融时间序列分析系列2

```python
这次讲一讲金融时间序列里面很重要的一类数据——收益率。因为一般来说，相对于价格数据，投资者可能更偏好收益率数据。然而收益率的定义有很多。这里我们以股票收益率为例来解释。
我们假设一支股票在t时刻价格为Pt,且股票在计算时间内没有分红。
```

## 1、收益率分类

### 1.1简单收益率
设t时刻股票价格为pt，t-1时刻股票价格为Pt-1。收益率为Rt.
       1 + Rt = Pt/Pt-1

```python
df =DataAPI.MktEqudGet(ticker=u"000001",beginDate=u"20170701",endDate=u"20170801",field=u"TradeDate,closePrice",pandas="1")
Rate=(df-df.shift(1))/df.shift(1)
Rt = Rate.dropna()
print Rt
```

### 1.2 年化收益率
设t时刻股票价格为pt，t-k时刻股票价格为Pt-k。收益率为Rt.这个期限为k年。
       1 + Rt = (Pt/Pt-k)^(1/k)

```python
import numpy as np 
import math
data =DataAPI.MktEqudGet(ticker=u"000001",beginDate=u"20170701",endDate=u"20170801",field=u"TradeDate,closePrice",pandas="1")
s0 = data["closePrice"][0]
s1 = np.array(data["closePrice"])[-1]
Rt= s1/s0
t = 365/len(data.index)
print "年化收益率为  %4.2f"%(math.pow(Rt,t)-1)
```

### 1.3 连续复合收益率
连续复合收益率也成为对数收益率。
设t时刻股票价格为pt，t-1k时刻股票价格为Pt-1。收益率为Rt。
       rt = ln(1 + Rt) = ln(Pt/Pt-1) = ln(Pt) - ln(Pt-1)
一般来说这个对数收益率用的最多，因为对数化具有可叠加性，好处理数据。

```python
import numpy as np 
import math
data =DataAPI.MktEqudGet(ticker=u"000001",beginDate=u"20170701",endDate=u"20170801",field=u"TradeDate,closePrice",pandas="1")
Rt = np.log(data["closePrice"]).diff().dropna()
print Rt
```

## 二、收益率的数学分布浅析

一般来说，数据的分布无非是看均值和方差。

### 2.1 正态分布

   正态分布（Normal distribution），也称“常态分布”，又名高斯分布（Gaussian distribution），最早由A.棣莫弗在求二项分布的渐近公式中得到。C.F.高斯在研究测量误差时从另一个角度导出了它。P.S.拉普拉斯和高斯研究了它的性质。
    正态曲线呈钟型，两头低，中间高，左右对称因其曲线呈钟形，因此人们又经常称之为钟形曲线。
    若随机变量X服从一个数学期望为μ、方差为σ^2的正态分布，记为N(μ，σ^2)。其概率密度函数为正态分布的期望值μ决定了其位置，其标准差σ决定了分布的幅度。当μ = 0,σ = 1时的正态分布是标准正态分布。


这里假设简单收益率服从正态分布。

```python
import scipy.stats as scs  
Rate=(df-df.shift(1))/df.shift(1)
Rt = np.array(Rate.dropna())

#正态性检验
#样本量小于50，用normal test 算法测试
p_value= scs.normaltest(Rt)[1]
if p_value<0.05:
    print "数据不符合正态分布"
else:
    print "数据符合正态分布。"
     
```

## 2.2 对数正态分布

如果 X 是服从正态分布的随机变量，则 exp(X) 服从对数正态分布；同样，如果 Y 服从对数正态分布，则 ln(Y) 服从正态分布。 如果一个变量可以看作是许多很小独立因子的乘积，则这个变量可以看作是对数正态分布。对数收益率就是服从对数正态分布

```python
import scipy.stats as scs  
import numpy as np 
import math
data =DataAPI.MktEqudGet(ticker=u"000001",beginDate=u"20170701",endDate=u"20170801",field=u"TradeDate,closePrice",pandas="1")
Rt = np.log(data["closePrice"]).dropna()
#正态性检验
#样本量小于50，用normal test 算法测试
p_value= scs.normaltest(Rt)[1]
if p_value<0.05:
    print "数据不符合正态分布"
else:
    print "数据符合正态分布。"
```

其他的分布还有就是稳定分布和正态分布的尺度混合。稳定分布其实就是正态分布的自然推广。而正态分布的尺度混合是指变量服从正态分布，然后方差服从另一个分布。这两个分布比较麻烦，以后遇到再讲讲。除了股票收益率，其实还有很多，这里主要是为了讲方法，其实都是一通万通的，掌握了方法就行。

具体在设计策略的时候，数据并不会有一个完美的分布，我们只能对其进行近似处理。
