```python
import numpy as np
import pandas as pd
from CAL.PyCAL import *
import matplotlib.pyplot as plt
import seaborn

def IN_OR_OUT(data,price,T):
    up=max(data['highPrice'][-T-1:-2])
    down=min(data['lowPrice'][-T-1:-2])
    if price>up:
        return 1
    elif price<down:
        return -1
    else:
        return 0
    
def CalcATR(data,T):
    TR_List=[]
    for i in range(1,T+1):
        TR=max(data['highPrice'][i]-data['lowPrice'][i], abs(data['highPrice'][i]-data['closePrice'][i-1]),abs(data['closePrice'][i-1]-data['lowPrice'][i]))
        TR_List.append(TR)
    ATR=np.array(TR_List).mean()
    return ATR

def Add_OR_Stop(price,lastprice,ATR,position_state):
    if (price>=lastprice+0.5*ATR and position_state==1) or (price<=lastprice-0.5*ATR        and position_state==-1):
        return 1
    elif (price<=lastprice-2*ATR and position_state==1) or (price<=lastprice+2*ATR            and position_state==-1):
        return -1
    else:
        return 0

def CheckPosition(position):
    try:
        if position.get('long_amount')==0 and position.get('short_amount')==0:
            return 0
        elif position.get('long_amount')>0:
            return 1
        elif position.get('short_amount')>0:
            return -1
        else:
            return 0
    except:
        return 0
    
def CheckPosition2(position):
    try:
        if position.get('long_position')==0 and position.get('short_position')==0:
            return 0
        elif position.get('long_position')>0:
            return 1
        elif position.get('short_position')>0:
            return -1
        else:
            return 0
    except:
        return 0
    
def CalcUnit(cash,ATR,coef):
    x=(cash*0.01/ATR)/coef
    return x
```
