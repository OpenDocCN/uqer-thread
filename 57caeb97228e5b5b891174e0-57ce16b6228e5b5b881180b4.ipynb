在学习完隐含马儿可夫模型后，想真正的了解下模型是否能够真正的预测股市，然而优矿的例子有点未来函数。于是重新理了下思路。
 从回测时间开始，每月的月末提取三年至当前的数据训练模型，然后利用模型每天计算市场状态，根据状态确定第二天的行动。结果比较一般，然后就这个原因进行了分析。

## 二、模型设置及状态意义获取
我在对中国的A股市场做了测试，测试时间从2010年4月1日至2014年1月1日（包含牛熊市场），依次来获得状态背后的意义。模型设置如下：
* 隐藏状态数（hidden state）：5（原来的例子是6个状态，但循环计算后会出现数据不足的问题，进行了降维处理）；
* 特征向量：一日对数收益差、五日对数收益差、当日对数高低价差、当日成交量、对数融资余额差，总计5个；
* 假设所有的特征向量的状态服从高斯分布（强假设，便于求解）

```python
#导入相关的模块
from hmmlearn.hmm import GaussianHMM
import datetime # python基本模块
import numpy as np
from matplotlib import cm, pyplot as plt
import matplotlib.dates as dates
import pandas as pd
import seaborn as sns
sns.set_style('white')
# 获取数据
beginDate = '20110401'
endDate = '20140101'

data = DataAPI.MktIdxdGet(ticker='000001',beginDate=beginDate,endDate=endDate,field=['tradeDate','closeIndex','lowestIndex','highestIndex','turnoverVol'],pandas="1")#1指数日行情数据
data1 = DataAPI.FstTotalGet(exchangeCD=u"XSHE",beginDate=beginDate,endDate=endDate,field=['tradeVal'],pandas="1")#1获取深圳交易所的融资融券数据
data2 = DataAPI.FstTotalGet(exchangeCD=u"XSHG",beginDate=beginDate,endDate=endDate,field=['tradeVal'],pandas="1")
tradeVal = data1 + data2 #1融资融券数据总和
tradeDate = pd.to_datetime(data['tradeDate'][5:])#日期列表
volume = data['turnoverVol'][5:]#2 成交量数据
closeIndex = data['closeIndex'] # 3 收盘价数据
deltaIndex = np.log(np.array(data['highestIndex'])) - np.log(np.array(data['lowestIndex'])) #3 当日对数高低价差
deltaIndex = deltaIndex[5:]
logReturn1 = np.array(np.diff(np.log(closeIndex))) #4 对数收益率
logReturn1 = logReturn1[4:]
logReturn5 = np.log(np.array(closeIndex[5:])) - np.log(np.array(closeIndex[:-5]))# 5日 对数收益差
logReturnFst = np.array(np.diff(np.log(tradeVal['tradeVal'])))[4:]
closeIndex = closeIndex[5:]
X = np.column_stack([logReturn1,logReturn5,deltaIndex,volume,logReturnFst]) # 将几个array合成一个2Darray
# Make an HMM instance and execute fit
model = GaussianHMM(n_components=5, covariance_type="diag", n_iter=1000).fit([X])
# Predict the optimal sequence of internal hidden state
hidden_states = model.predict(X)
#整合数据
res = pd.DataFrame({'tradeDate':tradeDate,'logReturn1':logReturn1,'logReturn5':logReturn5,'volume':volume,'state':hidden_states}).set_index('tradeDate')
plt.figure(figsize=(10, 6)) 
for i in range(model.n_components):
    idx = (hidden_states==i)
    idx = np.append(0,idx[:-1])#获得状态结果后第二天进行买入操作
    #fast factor backtest
    df = res.logReturn1
    res['sig_ret%s'%i] = df.multiply(idx,axis=0)    
    plt.plot(np.exp(res['sig_ret%s'%i].cumsum()),label='%dth hidden state'%i)
    plt.legend()
    plt.grid(1)
```

# 根据上图的结果，我们可以判断状态1和2比较明显的上升状态，以此为买入信号；而状态3明显的下跌状态，可以以此进行卖出操作；其他状态不操作

## 四、模型的不稳定性

出现上面的结果后，我思考既然隐含的状态背后意义需要我们自己确定，那么这个状态是否稳定呢？下面就进行简单的测试和分析，为了方便使用，先定义函数 HMMTest

```python
#导入相关的模块
from CAL.PyCAL import *
import pandas as pd
from hmmlearn.hmm import GaussianHMM
import datetime # python基本模块
import numpy as np
from matplotlib import cm, pyplot as plt
import matplotlib.dates as dates
import pandas as pd
import seaborn as sns
sns.set_style('white')

def HMMTest(beginDate,endDate):
	data = DataAPI.MktIdxdGet(ticker='000001',beginDate=beginDate,endDate=endDate,field=['tradeDate','closeIndex','lowestIndex','highestIndex','turnoverVol'],pandas="1")#1指数日行情数据
	data1 = DataAPI.FstTotalGet(exchangeCD=u"XSHE",beginDate=beginDate,endDate=endDate,field=['tradeVal'],pandas="1")#1获取深圳交易所的融资融券数据
	data2 = DataAPI.FstTotalGet(exchangeCD=u"XSHG",beginDate=beginDate,endDate=endDate,field=['tradeVal'],pandas="1")
	tradeVal = data1 + data2 #1融资融券数据总和
	tradeDate = pd.to_datetime(data['tradeDate'][5:])#日期列表
	volume = data['turnoverVol'][5:]#2 成交量数据
	closeIndex = data['closeIndex'] # 3 收盘价数据
	deltaIndex = np.log(np.array(data['highestIndex'])) - np.log(np.array(data['lowestIndex'])) #3 当日对数高低价差
	deltaIndex = deltaIndex[5:]
	logReturn1 = np.array(np.diff(np.log(closeIndex))) #4 对数收益率
	logReturn1 = logReturn1[4:]
	logReturn5 = np.log(np.array(closeIndex[5:])) - np.log(np.array(closeIndex[:-5]))# 5日 对数收益差
	logReturnFst = np.array(np.diff(np.log(tradeVal['tradeVal'])))[4:]
	closeIndex = closeIndex[5:]
	X = np.column_stack([logReturn1,logReturn5,deltaIndex,volume,logReturnFst]) # 将几个array合成一个2Darray
	# Make an HMM instance and execute fit
	model = GaussianHMM(n_components=5, covariance_type="diag", n_iter=1000).fit([X])
	# Predict the optimal sequence of internal hidden state
	hidden_states = model.predict(X)
	#整合数据
	res = pd.DataFrame({'tradeDate':tradeDate,'logReturn1':logReturn1,'logReturn5':logReturn5,'volume':volume,'state':hidden_states}).set_index('tradeDate')
	plt.figure(figsize=(10, 6)) 
	for i in range(model.n_components):
		idx = (hidden_states==i)
		idx = np.append(0,idx[:-1])#获得状态结果后第二天进行买入操作
		#fast factor backtest
		df = res.logReturn1
		res['sig_ret%s'%i] = df.multiply(idx,axis=0)
		plt.plot(np.exp(res['sig_ret%s'%i].cumsum()),label='%dth hidden state'%i)
		plt.legend()
		plt.grid(1)
	#print res
```

```python
beginDate = '20100401'
endDate = '20140101'
HMMTest(beginDate,endDate)
beginDate = '20100401'
endDate = '20140301'
HMMTest(beginDate,endDate)
```

*数据长度变了两个月，训练出来的模型就完全不一样，完全不具备稳定性，这样用这个模型确实有点后怕。那么问题出在什么地方呢？
1、可能输入数据不足或者因子效果不够。一日对数收益差、五日对数收益差、当日对数高低价差、当日成交量、对数融资余额差，这五个数据从感觉上也不足以预测市场的状态；
2、可以考虑每个月更新模型后同时计算状态的阶段收益，动态的识别状态；
3、看用于训练隐含马尔科夫模型的鲍姆-韦尔奇算法介绍，算法能收敛到一个局部最优点，很遗憾它一般不能保证找到全局最优点。这样数据或者因子有效性不足时收敛的结果可能差别会比较大，即便同样的数据，多训练几次，可能结果就不一样，结果可靠性又打了折扣。
看来要应用这个模型，不仅需要找好因子，还需要深层次思考内在机制，提高系统稳定性和可靠性！*
