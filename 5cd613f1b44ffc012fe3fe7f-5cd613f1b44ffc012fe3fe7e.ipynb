**马克维茨投资组合理论的基本假设为：**
--
* 投资者是风险规避的，追求期望效用最大化；
* 投资者根据收益率的期望值与方差来选择投资组合；
* 所有投资者处于同一单期投资期。马克维茨提出了以期望收益及其方差($E，δ^2$)确定有效投资组合。

以期望收益$E$来衡量证券收益，以收益的方差$δ^2$表示投资风险。资产组合的总收益用各个资产预期收益的加权平均值表示，组合资产的风险用收益的方差或标准差表示，则马克维茨优化模型如下：
　　$$\min&emsp;\delta^2(r_p)=\sum\sum w_iw_jcov(r_i,r_j) = w^T∑_{cov}w$$
    $$s.t. &emsp; -w < 0 $$
	$$sum(w)=1$$
　　$$  E(r_p)=\sum w_ir_i$$

　　式中：
  $r_p$——组合收益；

  $r_i$、$r_j$——第i种、第j种资产的收益；

  $w_i$、$w_j$——资产i和资产j在组合中的权重；

  $δ^2(r_p)$——组合收益的方差即组合的总体风险；

  $cov(r_i,r_j)$——两种资产收益率之间的协方差。
  $∑_{cov}$——资产组合的收益率协方差矩阵
  
从经济学的角度分析，就是说投资者预先确定一个期望收益率，然后通过$E(r_p)=\sum w_ir_i$确定投资组合中每种资产的权重，使其总体投资风险最小，所以在不同的期望收益水平下，得到相应的使方差最小的资产组合解，这些解构成了最小方差组合，也就是我们通常所说的**有效组合**。有效组合的收益率期望和相应的最小方差之间所形成的曲线，就是**有效组合投资的前沿**。投资者根据自身的收益目标和风险偏好，在有效组合前沿上选择最优的投资组合方案。**有效前沿优化函数：**$\min&emsp;portfolios =\delta^2(r_p) - E(r_p)$，在指定的收益下选择风险最小的组合，或者说在指定的风险下选择收益最大的组合。


```python
import numpy as np
import matplotlib.pyplot as plt
import cvxopt as opt
from cvxopt import blas, solvers
import pandas as pd

# 关掉进度展示，进度展示是运行过程进度的一个打印输出，可以通过其查看代码运行进度
solvers.options['show_progress'] = False
```

```python
# 有效前沿曲线展示

# 产生随机权重的函数
def rand_weights(n):
    ''' Produces n random weights that sum to 1 '''
    k = np.random.rand(n)
    return k / sum(k)

# 返回组合收益率和波动性
def random_portfolio(returns):

    p = np.asmatrix(np.mean(returns, axis=1))
    w = np.asmatrix(rand_weights(returns.shape[0]))
    C = np.asmatrix(np.cov(returns))
    
    mu = w * p.T
    sigma = np.sqrt(w * C * w.T)
    
    # 过滤器
    if sigma > 2:
        return random_portfolio(returns)
    return mu, sigma

# 获取有效前沿曲线
def optimal_portfolio(returns):
    n = len(returns)
    returns = np.asmatrix(returns)
    
    N = 100
    mus = [10**(5.0 * t/N - 1.0) for t in range(N)]   
    # mus在[0.1,8912]，给标准差配置不同的权重，获得不同的前沿点，权重越大，越靠近标准差最小点，即子弹头
    
    # 转化为cvxopt matrices
    S = opt.matrix(np.cov(returns))
    pbar = opt.matrix(np.mean(returns, axis=1))
    
    # 约束条件
    G = -opt.matrix(np.eye(n))   # 组合权重大于0
    h = opt.matrix(0.0, (n ,1))
    A = opt.matrix(1.0, (1, n))  # 组合权重为1
    b = opt.matrix(1.0)
    
    # 使用凸优化计算有效前沿
    # 指定不同的方差来获得最大收益，即能得到有效前沿曲线
    portfolios = [solvers.qp(mu*S, -pbar, G, h, A, b)['x'] 
                  for mu in mus]
    
    ## 计算有效前沿的收益率和风险
    returns = [blas.dot(pbar, x) for x in portfolios]
    risks = [np.sqrt(blas.dot(x, S*x)) for x in portfolios]
    m1 = np.polyfit(returns, risks, 2)  
    func = np.poly1d(m1)
    
    # 选择组合
    # x1 = np.sqrt(m1[2] / m1[0])   # 为什么使用这样的x1，这样定义的最优组合肯定是选择收益率最大的weight
    x1 = 100   # 根据牛熊市来调整x1的值
    wt = solvers.qp(opt.matrix(x1 * S), -pbar, G, h, A, b)['x']
    choice_return = blas.dot(pbar, wt)
    choice_risk = np.sqrt(blas.dot(wt, S*wt))
    return returns, risks, func,choice_return,choice_risk, np.asarray(wt)


np.random.seed(1)

# 随机生成数据
## N资产数量
n_assets = 4
## 收益率长度
n_obs = 1000
return_vec = np.random.randn(n_assets, n_obs)

# 真实股票数据
start = '2010-09-02'                     
end = '2011-04-30'                       
secids = [ '600036.XSHG', '600030.XSHG','601318.XSHG', '600000.XSHG']
df_dailyReturn = DataAPI.EquRetudGet(secID=secids,listStatusCD=u"",beginDate=start,endDate=end,isChgPctl=u"",field=u"secID,tradeDate,dailyReturnNoReinv",pandas="1")
df_dailyReturn = df_dailyReturn.dropna()
df_dailyReturn=df_dailyReturn.pivot_table(index='secID',columns='tradeDate',values='dailyReturnNoReinv').sort_index(axis=1,ascending=True)
returns_test = df_dailyReturn.iloc[:,:12].values
return_vec = returns_test

# 随机组合2000次，看收益率与标准差分布情况
n_portfolios = 2000
means, stds = np.column_stack([random_portfolio(return_vec) for _ in range(n_portfolios)])

# 获取有效组合前沿曲线，根据牛熊市选择收益率和风险情况
returns, risks ,func,choice_return,choice_risk,weights = optimal_portfolio(return_vec)
print weights
print returns
print risks
print choice_return
print choice_risk

# xmin ,xmax = stds.min(), stds.max()
# ymin, ymax = means.min(), means.max()
# dx = (xmax - xmin) * 0.1
# dy = (ymax - ymin) * 0.1
# plt.xlim(xmin - dx, xmax + dx)
# plt.ylim(ymin - dy, ymax + dy)

plt.xlabel('std')    # 标准差-波动性
plt.ylabel('mean')   # 平均值-收益率
plt.title('Mean and standard deviation of returns of randomly generated portfolios') # 每个投资组合的收益率和波动性的散点图

plt.plot(stds, means, 'o', markersize=2)
plt.plot(risks, returns, 'r-*')
# plt.plot(func(returns), returns, 'b-o')
plt.plot(choice_risk, choice_return, 'g-*', markersize=20)
plt.show()
```

并根据此权重比例进行下单和调仓。
