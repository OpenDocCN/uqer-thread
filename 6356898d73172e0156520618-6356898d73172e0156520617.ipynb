# 逻辑回归预测股价

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn import svm
```

```python
data=DataAPI.MktEqudGet(secID=u"",ticker=u"600235",tradeDate=u"",beginDate=u"20200101",endDate=u"20221010",isOpen=""
                               , field=u"",pandas="1")
data.head()
```

```python
data.set_index('tradeDate',inplace=True)
data.index=pd.to_datetime(data.index)
```

```python
data.head()
```

```python
data.dropna(inplace=True)
```

特征工程

```python
for i in range(1, 8, 1):
    data['closePrice - ' + str(i) + 'd'] = data['closePrice'].shift(i)     #收盘价closePrice
```

```python
for i in range(1, 8, 1):
    data['turnoverVol - ' + str(i) + 'd'] = data['turnoverVol'].shift(i)    #成交量turnoverVol
```

```python
for i in range(1, 8, 1):
    data['turnoverRate - ' + str(i) + 'd'] = data['turnoverRate'].shift(i)   #换手率turnoverRate
```

```python
for i in range(1, 8, 1):
    data['PE - ' + str(i) + 'd'] = data['PE'].shift(i)                #滚动市盈率PE
```

```python
for i in range(1, 8, 1):
    data['PB - ' + str(i) + 'd'] = data['PB'].shift(i)      #市净率PB
```

```python
data.head(8)
```

```python
hs_7d_1= data[[x for x in data.columns if 'closePrice' in x]].iloc[7:]
hs_7d_1.head()
```

```python
hs_7d_2= data[[x for x in data.columns if 'turnoverVol' in x]].iloc[7:]
```

```python
hs_7d_3= data[[x for x in data.columns if 'turnoverRate' in x]].iloc[7:]
```

```python
hs_7d_4= data[[x for x in data.columns if 'PE' in x]].iloc[7:]
```

```python
hs_7d_5= data[[x for x in data.columns if 'PB' in x]].iloc[7:]
```

```python
hs_7d_12= pd.merge(hs_7d_1, hs_7d_2, left_index=True, right_index=True)
```

```python
hs_7d_34= pd.merge(hs_7d_3, hs_7d_4, left_index=True, right_index=True)
```

```python
hs_7d_1234= pd.merge(hs_7d_12, hs_7d_34, left_index=True, right_index=True)
```

```python
hs_7d = pd.merge(hs_7d_1234, hs_7d_5, left_index=True, right_index=True)
```

```python
hs_7d.head()
```

```python
import sklearn
from sklearn import linear_model
```

```python
X_train = hs_7d
X_train = sklearn.preprocessing.scale(X_train)    #对features进行标准化；
X_train
```

```python
lm = linear_model.LogisticRegression(C=1000)
```

```python
y_train = np.sign(hs_7d['closePrice'].pct_change().shift(-1))       #非常重要；拿到下一天的收益，用.shift(-1)
y_train.replace(to_replace= np.NaN, value = 0,inplace = True)
y_train = y_train.values.reshape(-1,1)
```

```python
lm.fit(X_train, y_train)
```

```python
lm.score(X_train, y_train)
```

```python
data['prediction'] = np.NaN
data['prediction'].iloc[7:] = lm.predict(X_train) 
```

```python
data['prediction'].value_counts()
```

```python
data['strategy'] = (data['prediction'].shift(1) * data['chgPct'] + 1).cumprod()
data['cum_ret'] = (data['chgPct']+1).cumprod()      #chgPct涨跌幅
```

```python
data[['strategy','cum_ret']].dropna().plot(figsize=(10, 6))
```
