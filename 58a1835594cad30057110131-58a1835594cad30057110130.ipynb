## 总结
---
从历史上两会前后的整体的表现上来看，大盘及行业板块在两会前后存在较为明显的日历效应，两会前20个交易日至两会开始，市场整体上涨概率高，平均涨跌幅为正，但临近两会或两会期间，大盘平均涨跌幅为负，且上涨概率低于50%；而会议结束后，5到20个交易日上涨概率与平均涨跌幅均较高。两会前市场的整体风格表现较为明显，中小板的表现要优于沪深300的表现。
从行业上看，两会开始前20日，益多配置综合、有色金属、农林牧渔等行业。而临近两会或者在两会期间，可以适当降低仓位，配一些前期涨幅较低的行业。两会结束的5日至20日，益多配置房地产、轻工制造、休闲服务等行业。

## 附日历效应模板及画图函数

```python
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter 
from CAL.PyCAL import *    # CAL.PyCAL中包含font
font.set_size(15)
def get_confer_plot(sec_ret,sec_ratio,legend=[]):
    sec_ret.index.name = None
    sec_ratio.index.name = None
    fig = plt.figure(figsize=(12,6))
    ax = fig.add_subplot(111)
    ax2=ax.twinx()
    sec_ret.sort_values(ascending=False).plot(kind='bar',color = ['red'],ax=ax)
    ax2.plot(sec_ratio.ix[list(sec_ret.sort_values(ascending=False).index)].values,'blue',linewidth=3.5)
    def myfunc(x, pos=0): 
        return "{0:.2f}%".format(x * 100)
    ax.yaxis.set_major_formatter(FuncFormatter(myfunc))
    ax2.yaxis.set_major_formatter(FuncFormatter(myfunc))
    ax.set_xticklabels([name.decode('utf-8') for name in list(sec_ret.sort_values(ascending=False).index)],fontproperties=font)
    ax.legend([legend[0]],loc='upper left',bbox_to_anchor=(0.08,1.1),prop=font,frameon= False,handlelength = 3,handletextpad=0.1,borderpad = 0.1)
    ax2.legend([legend[1]],loc='upper left',bbox_to_anchor=(0.52,1.1),prop=font,frameon= False,handlelength = 3,handletextpad=0.1,borderpad = 0.1)
    ax.spines['top'].set_color('none')
    ax2.spines['top'].set_color('none')
    ax.xaxis.set_ticks_position('bottom')
    plt.show()
```

```python
import numpy as np
import pandas as pd
import datetime
from dateutil.parser import parse

class EventQuotation:

    def __init__(self,event_date,event_name):
        self.event_name = event_name
        self.event = pd.DataFrame(event_date).rename(columns={0:self.event_name+'开始时间',1:self.event_name+'结束时间'})
        self.trade_date = None
        self.df_index = None
        
    def update_trade_date(self):
        """
        更新获取交易日历
        """
        trade_cal = DataAPI.TradeCalGet(exchangeCD=u"XSHG",beginDate=u"",endDate=u"",field=u"isOpen,calendarDate",pandas="1")
        trade_cal = trade_cal[trade_cal['isOpen']==1]
        self.trade_date = trade_cal['calendarDate'].map(lambda x:parse(x)).tolist()
        
    def get_index_equotation(self,index_symbol):
        """
        获取指数行情
        """
        df_index = DataAPI.JY.MktIdxdJYGet(beginTime=u"",endTime=u"",indexID=index_symbol,secIDSrc=u"",
                                           startDate=u"",endDate=u"",field=u"tradeDate,closeIndex",pandas="1")
        df_index['tradeDate'] = df_index['tradeDate'].apply(lambda x:x[0:10])
        self.df_index = df_index.set_index('tradeDate')['closeIndex']
    
    @staticmethod
    def step_trade_date(base_date, step, trading_days):
        """
        交易日向前向后漂移
        """
        base_date_index = trading_days.index(parse(base_date))
        step_date_index = base_date_index + step
        return trading_days[step_date_index].strftime('%Y-%m-%d')
    
    @staticmethod
    def adjust_to_trade_date(base_date, trading_days, adjust='last'):
        """
        调整基准日，若基准日不是交易日，则将其调整为之前最近的一个交易日
        """
        base_date = base_date.to_datetime() if isinstance(base_date, pd.tslib._Timestamp) else base_date
        base_date = parse(base_date.strftime('%Y%m%d')) if isinstance(base_date, datetime.datetime) else parse(base_date)
        while base_date not in trading_days:
            if adjust == 'last':
                base_date = base_date - datetime.timedelta(days=1)
            elif adjust == 'next':
                base_date = base_date + datetime.timedelta(days=1)
            if base_date < datetime.datetime(1990, 11, 3):
                raise ValueError('input date can not be earlier than 1990-11-03')
            elif base_date > datetime.datetime.today():
                raise ValueError('BASE_DATE can not be later than today')
        return base_date.strftime('%Y-%m-%d')
    
    def get_start_end_yield(self,confer_date):
        """
        获取事件区间的收益率
        """
        confer_date = confer_date.copy()
        confer_date['start_adjust_date'] = confer_date[self.event_name+'开始时间'].map(lambda x:self.step_trade_date(self.adjust_to_trade_date(x,self.trade_date,'next'),-1,self.trade_date))
        confer_date['end_adjust_date'] = confer_date[self.event_name+'结束时间'].map(lambda x:self.adjust_to_trade_date(x,self.trade_date,'last'))
        confer_date['start_adjust_date_index'] = confer_date['start_adjust_date'].map(lambda x:self.df_index.ix[x])
        confer_date['end_adjust_date_index'] = confer_date['end_adjust_date'].map(lambda x:self.df_index.ix[x])
        
        return (confer_date['end_adjust_date_index'] - confer_date['start_adjust_date_index'])/confer_date['start_adjust_date_index']
    
    def get_step_window_yield(self,confer_date,step):
        """
        获取事件开始前多少天到事件开始的收益，或事件结束到事件结束后多少天的收益
        """
        confer_date = confer_date.copy()
        if step < 0:
            confer_date['adjust_date'] = confer_date[self.event_name+'开始时间'].map(lambda x:self.step_trade_date(self.adjust_to_trade_date(x,self.trade_date,'next'),-1,self.trade_date))
        else:
            confer_date['adjust_date'] = confer_date[self.event_name+'结束时间'].map(lambda x:self.adjust_to_trade_date(x,self.trade_date,'last'))

        confer_date['step_adjust_date'] = confer_date['adjust_date'].map(lambda x:self.step_trade_date(x,step,self.trade_date))
        confer_date['adjust_date_indx'] = confer_date['adjust_date'].map(lambda x:self.df_index.ix[x])
        confer_date['step_adjust_date_index'] = confer_date['step_adjust_date'].map(lambda x:self.df_index.ix[x])
        if step > 0:
            return (confer_date['step_adjust_date_index'] - confer_date['adjust_date_indx'])/confer_date['adjust_date_indx']
        else:
            return (confer_date['adjust_date_indx'] - confer_date['step_adjust_date_index'])/confer_date['step_adjust_date_index']
    
    def get_summary_df(self,static_list,index_symbol,interval=False):
        """
        汇总收益列表
        """
        self.get_index_equotation(index_symbol)
        
        if interval == True:
            self.event[self.event_name+'期间涨跌幅'] = self.get_start_end_yield(self.event)
            self.event[self.event_name+'期间涨跌幅'] = pd.Series(["{0:.2f}%".format(ret * 100) for ret in self.event[self.event_name+'期间涨跌幅']],index = self.event.index)
            
        for step in static_list:
            if step < 0:
                self.event[self.event_name+'前'+str(-step)+'交易日涨跌幅'] = self.get_step_window_yield(self.event,step)
                self.event[self.event_name+'前'+str(-step)+'交易日涨跌幅'] = pd.Series(["{0:.2f}%".format(ret * 100) for ret in self.event[self.event_name+'前'+str(-step)+'交易日涨跌幅']],index = self.event.index)
            elif step > 0:
                self.event[self.event_name+'后'+str(step)+'交易日涨跌幅'] = self.get_step_window_yield(self.event,step)
                self.event[self.event_name+'后'+str(step)+'交易日涨跌幅'] = pd.Series(["{0:.2f}%".format(ret * 100) for ret in self.event[self.event_name+'后'+str(step)+'交易日涨跌幅']],index = self.event.index)
            else:
                raise ValueError('step can not be zero!')
        return self.event
    
    def get_static_df(self,summary_df):
        """
        统计汇总列表的统计特征
        """
        summary_df = summary_df.copy()
        summary_df = summary_df.set_index([self.event_name+'开始时间',self.event_name+'结束时间'])
        summary_df = summary_df.applymap(lambda x:float(x.strip('%'))/100)
        summary_df.columns = summary_df.columns.map(lambda x:x.replace('涨跌幅',''))
        static_df = pd.concat([summary_df.mean(),summary_df.apply(lambda x:1.0*len(x[x>0])/len(x),axis=0)],axis=1).T
        static_df = static_df.rename(index={0:'平均涨跌幅',1:'上涨概率'})
        static_df = static_df.applymap(lambda x:"{0:.2f}%".format(x * 100))
        return static_df
    
    def get_mutil_symbol_static(self,static_list,index_list,interval=False):
        """
        """
        avg_return = pd.DataFrame()
        up_ratio = pd.DataFrame()
        for symbol,name in index_list:
            summary_df = self.get_summary_df(static_list,symbol,interval=interval)
            summary_df = summary_df.copy()
            summary_df = summary_df.set_index([self.event_name+'开始时间',self.event_name+'结束时间'])
            summary_df = summary_df.applymap(lambda x:float(x.strip('%'))/100)
            summary_df.columns = summary_df.columns.map(lambda x:x.replace('涨跌幅',''))
            static_df = pd.concat([summary_df.mean(),summary_df.apply(lambda x:1.0*len(x[x>0])/len(x),axis=0)],axis=1).T
            avg_return = pd.concat([avg_return,static_df.iloc[[0]].rename(index={0:name})],axis=0)
            up_ratio = pd.concat([up_ratio,static_df.iloc[[1]].rename(index={1:name})],axis=0)
        avg_return = avg_return.applymap(lambda x:"{0:.2f}%".format(x * 100))
        up_ratio = up_ratio.applymap(lambda x:"{0:.2f}%".format(x * 100))
        avg_return.index.name = '平均涨跌幅'
        up_ratio.index.name = '平均上涨比例'
        return avg_return,up_ratio
```
