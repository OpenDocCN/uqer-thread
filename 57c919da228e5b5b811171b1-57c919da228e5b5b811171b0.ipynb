```python
import time
import pandas as pd
import numpy as np

# data from api
# P: exchange code? XDCE,XSGE,XZCE
def futMain():
    # get data from api
    data = DataAPI.FutuGet(exchangeCD=[u"XSGE",u"XZCE",u"XDCE"],secID=u"",ticker=u"",contractStatus="L",contractObject=u""
                           ,field=['exchangeCD','contractObject','minChgPriceNum',u'minChgPriceUnit',u'limitUpNum',u'contMultNum'],pandas="1")
    data1 = data.drop_duplicates(subset = ['exchangeCD','contractObject'])
    data1.sort(['exchangeCD','contractObject'])
    return data1

def trdDayCal():
    # format of %Y-%m-%d
    data = DataAPI.TradeCalGet(exchangeCD=u"XSHG,XSHE",beginDate=u"20100101",endDate=u"",field=u"",pandas="1")
    data1 = data.calendarDate[data.isOpen == 1].tolist()
    return(data1)

#**** data:
# futM
futM = futMain()
# trdDayCal()[-4:-1]
tdates = trdDayCal()

tdate = pd.datetime(2016, 8, 26).strftime('%Y-%m-%d')
fdate = tdates[tdates.index(tdate) - 100]

```

一 商品46品种简介
    1. 整体合约的大小在2——6万，以10%保证金估算，一手的保证金主要在2000——6000；
    2. 成交量最活跃的品种，RB、M、I、RM等，都是合约金额较小的品种，投机者偏爱小合约；
    3. 大合约：黄金28万，铜18万，橡胶和焦炭12万。
    4. 过滤的品种15个（主力合约成交量在10000以下，尽量避免）：
        B、BB、FB、FU、JR、LR、PB、PM、RI、RS、SF、SM、SN、WH、WR

```python

def contSize(tdate,tdates):

    data = DataAPI.MktMFutdGet(tradeDate=tdate,mainCon=1,contractMark=u"",contractObject=u"",startDate=u"",endDate=u""
                               ,field=[u"contractObject",u"exchangeCD",u"tradeDate",u"closePrice",u"turnoverVol"],pandas="1") # [u"ticker",
    # print data.head(2)
    # print data.contractObject[data.turnoverVol <= 10000][data.exchangeCD != u'CCFX']
    data = data[data.turnoverVol > 10000 ][data.exchangeCD != u'CCFX'] # 

    fut = futMain()
    data = data.merge(fut[['contractObject','minChgPriceNum','contMultNum']],on = 'contractObject')
    data['value'] = data['closePrice']*data['contMultNum']/10000
    data['volCap'] = data['closePrice']*data['contMultNum']*data['turnoverVol']/100000000
    data = data.sort('turnoverVol',ascending = False)

    # 画图:main contract
    data.plot(x = ['contractObject'],y=['value'], kind='bar', figsize=(14, 7), title=u'cnotractSize/1e4', color='g')

contSize(tdate,tdates)

```

```python
def mktSize(date):
    data = DataAPI.MktFutdGet(tradeDate=date,secID=u"",ticker=u"",beginDate=u"",endDate=u""
                              ,field=[u"ticker",u'contractObject',u"closePrice",u"turnoverVol",u"turnoverValue",u"openInt"],pandas="1")
    # print data.head(2)
    data['oiCap'] = data.closePrice*data.openInt/1e8
    data.turnoverValue /= 1e8
    # data['volCap'] = data.closePrice*data.turnoverVol
    
    data1 = data[['contractObject',u"turnoverVol","turnoverValue","openInt",'oiCap']].groupby(by = "contractObject").sum()
    fut = futMain()
    data1 = data1.merge(fut[['contractObject','minChgPriceNum','contMultNum']],left_index = True, right_on = 'contractObject')
    data1.oiCap *= data1.contMultNum
    
    return data1

mktData = mktSize(tdate).sort("turnoverValue",ascending = False)
# data = DataAPI.MktFutdGet(tradeDate=date,secID=u"",ticker=u"",beginDate=u"",endDate=u"",field=u"",pandas="1")
# data.head(2)
mktData['spec'] = mktData.turnoverValue/mktData.oiCap

```

二 市场容量，成交持仓，资金流向统计，投机度等
    1. 成交金额和持仓金额对比，投机性强弱；
    2. 大合约：黄金和铜，市场容量和螺纹相近，投机明显少，除基本面，也说明投机者偏向小合约；
    3. 螺纹钢和铁矿成交金额最大，同时市场容量较大，说明黑色热度很高；

```python
# plot2: total cap
mktData.plot(x = ['contractObject'],y=['turnoverValue','oiCap'], kind='bar', figsize=(14, 7), title=u'volCap&oiCap/1e8', color=['g','r'])

mktData.plot(x = ['contractObject'],y=['spec'], kind='bar', figsize=(14, 7), title=u'speculation', color=['r'])

mktData[mktData.turnoverVol > 10000].plot(y=['oiCap'], kind='pie', figsize=(10, 10),labels = mktData.contractObject)
mktData[mktData.turnoverVol > 10000].plot(y=['turnoverValue'], kind='pie', figsize=(10, 10),labels = mktData.contractObject)
mktData.head()
```

三 主力合约相关性
    1. 日收益标准差：1%-3%之间，超过2.5%的分别是HC,I,J,JM,FB都是黑色品种。
    2. 相关性：整体商品有一定正相关性（系统风险较大）；
    3. 特定分组（如AU和AG，CU和ZN，RB和J等）内部相关性很高；可以为配对交易提供信息。

```python
# retData & cor
def corFut(fdate,tdate):
    # retData = DataAPI.MktMFutdGet(tradeDate=u"",mainCon=1,contractMark=u"",contractObject=u"",startDate=fdate,endDate=tdate ,field=[u""],pandas="1")
    data = DataAPI.MktMFutdGet(tradeDate=u"",mainCon=1,contractMark=u"",contractObject=u"",startDate=fdate,endDate=tdate
                                  ,field=[u"ticker",u"exchangeCD",u'tradeDate', u'contractObject'
                                          , u'preClosePrice', u'closePrice', u'turnoverVol'],pandas="1")
    data = data[data.exchangeCD != u'CCFX']
    data['ret'] = data.closePrice/data.preClosePrice - 1
    
    # filter of contractObject
    data1 = data[['contractObject','turnoverVol']].groupby('contractObject').mean()
    contractObs = data1[data1.turnoverVol > 10000].index.tolist()
    # return data1
    # contractObs = data.contractObject.unique()
    # print contractObs
    dataRet = pd.DataFrame([],index = data.tradeDate.unique(),columns=contractObs)
    #*** pivot_table
    for i in contractObs:
        dat = data.ret[data.contractObject == i]
        #print i,type(dat)
        dat.index = data.tradeDate[data.contractObject == i]
        #print dat
        dataRet[i] = dat
    #*** table: cor
    sampleCon = ['AU','AG','CU','ZN','RB','J','RM','M','Y','P','L','RU']
    print dataRet.corr().round(3).loc[sampleCon,sampleCon]
    
    #*** plot: sd
    data2 = np.sqrt(np.diag(dataRet.cov()))*100
    pd.DataFrame({'cont': dataRet.cov().columns,'sd': data2}).plot(x = ['cont'],y=['sd'], kind='bar', title=u'sd(%)', color=['r'])
    #return dataRet

corFut(fdate,tdate)
# data.head()
```
