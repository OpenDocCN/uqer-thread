感谢Myron：https://uqer.datayes.com/community/share/573058f3228e5b303767f191
但是我个人感觉他写的逻辑有问题：
问题在于operaPipeline在startPipeline没有达到买入卖出启动的时候就开始计数了；
我修改了一下，应该逻辑正常了
但是效果觉得很差，应该比不上海龟的择时效果
我思考了一下原因：
GFTD本质应该是一种反转策略，这个策略是这样定义反转的：
当close价格连续下跌后n2根bar，开始买入启动，开始计算符合条件的bar，如果bar数量达到一定程度，就开多；
反之亦然；
但是并不是每次趋势结束，都是以close价格连续下跌n2根bar收尾的；
当一次反转没有抓住，而且没有设置止损的话，将是大幅的亏损
