# 问题3. 为什么很多情况下没有发现前复权的未来信息影响？



### 我们先来思考最开始的那个策略是如何产生未来信息的影响的。既然get_history()和DataAPI.MktEqudAdjGet的数据一样，我们来用DataAPI.MktEqudAdjGet的数据计算下vwap

```python
data = DataAPI.MktEqudAdjGet(secID=u"",ticker=u"000001",beginDate=u"20120101",endDate=u"20160801",isOpen="",field=u"",pandas="1")  
data = data.set_index('tradeDate')
vwap = data['turnoverValue']/data['turnoverVol']

data['closePrice'].plot(figsize=(20,8),label='adjPrice')
vwap.plot(label='vwap')
plt.legend(loc=0)
```

### 我们发现，**vwap是十分近似真实价格（未复权）的**。 根据本帖一开始的策略，可以表达为：
$$ \large vwap-\sqrt{high*low}$$

### 我们知道前复权的未来信息在于基于**当前价格**进行复权，如果我们的计算只涉及前复权价格参考系，这个未来信息其实没有用到；但是一旦我们和真实价格参考系进行了计算（上式里为求差），那未来信息就暴露了，**因为我们的结果含有了回测时刻价格相对当前价格的差。** 

### 因此我推测，如果只使用前复权数据，不和真实数据产生对比，未来信息未必会暴露。

## 所以，以后大家用到 **DataAPI.MktEqudAdjGet 、 get_history 、 快速signal** 的时候得注意了，里面数据的计算要尽量避免和真实数据混合计算而导致未来信息的暴露。
