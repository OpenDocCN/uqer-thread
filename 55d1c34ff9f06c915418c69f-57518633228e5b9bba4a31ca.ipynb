```python
# 需要将上面字典的中文名换成ticker
# 从DataAPI里找到一个可以查询对应关系的API
df = DataAPI.SecIDGet(assetClass=u"E",ticker=u"",partyID=u"",cnSpell=u"",field=u"",pandas="1")
# 过滤掉非沪深以及非上市状态的股票
df = df[(df.exchangeCD.str.startswith('XSH'))&(df.listStatusCD=='L')]
df.head()
```

```python
# 设置到universe里
universe = {}
for k in guo_gai:
    universe[k] = []
    for stk in guo_gai[k].split('、'):
        universe[k].extend(df[df.secShortName==stk].ticker.values)
```

```python
# 计算间隔日期的涨跌幅
import pandas as pd
result = pd.DataFrame(columns=['地域', '涨跌幅'])
start_date = '20150814'
end_date = '20150817'
for region in universe:
    df = DataAPI.MktEqudAdjGet(ticker=','.join(universe[region]), beginDate=start_date, endDate=end_date)
    result.loc[len(result)] = region, (df.groupby('ticker')['closePrice'].last()/df.groupby('ticker')['closePrice'].first()).mean()
```
