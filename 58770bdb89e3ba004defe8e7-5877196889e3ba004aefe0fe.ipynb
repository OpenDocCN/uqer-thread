# `准备数据`

- 基本行情
- 前瞻收益率
- 历史收益率

```python
df = DataAPI.MktEqudAdjGet(secID=set_universe('A'), beginDate=u"20160101", endDate=u"20161231", 
                           field=['ticker', 'secID', 'tradeDate', 'preClosePrice', 'openPrice', 'highestPrice', 
                                  'lowestPrice', 'closePrice', 'marketValue', 'turnoverRate', 'isOpen'])
df['change'] = df.closePrice / df.preClosePrice - 1.0

### 前瞻，历史N日收益率
def forward_history_return(df, window=10):
    """
    HR: history return
    FR: forward return
    """
    df['HR_{}'.format(window)] = df.closePrice / df.closePrice.shift(window) - 1.0
    df['FR_{}'.format(window)] = df.closePrice.shift(-1 * window) / df.closePrice - 1.0
    return df

df = df.groupby(by='ticker').apply(func = lambda x: forward_history_return(x, 10))

df.head()
```

# `获取需要分析的因子`

- DataAPI 中包含的因子
- 自己用行情数据计算的其他因子

```python
### 获取区间因子，自己计算，或者直接使用 DataAPI 中提供的因子
def factor_get(df, tstart, tend, factors, filename):
    """获取多只股票多日因子数据
    """
    df2 = pd.DataFrame()
    for i in df.ticker.unique().tolist():
        d = DataAPI.MktStockFactorsDateRangeGet(ticker=i, beginDate=tstart, endDate=tend,
                                                field=['ticker', 'tradeDate'] + factors)
        df2 = df2.append(d)
        
    df2 = df2.reset_index(drop=True)
    df = df.merge(df2, on=['ticker', 'tradeDate'])
    df = df.set_index('tradeDate')
    
    df.to_csv(filename)
    
    return df

df = factor_get(df, '20160101', '20161231', [factor_name], filename='Factor_Analysis_{}.csv'.format(factor_name))
df.head()
```

#  `因子预测能力初探`

- 计算了每一天的因子和10日前瞻收益率的秩相关系数【某一天的因子数据，某一天的10日收益】

```python
def calculate_factor_ic(df, factor):
    df_ic = pd.DataFrame(index=df.index.unique(), columns=['IC','pValue'])

    for day in df.index.unique():
        factor_return = df.ix[day][[factor, 'FR_10']]
        factor_return = factor_return[(~np.isnan(factor_return[factor])) & (~np.isnan(factor_return.FR_10))]
        ic, p_value = st.spearmanr(factor_return[factor], factor_return['FR_10'])
        df_ic['IC'][day] = ic
        df_ic['pValue'][day] = p_value

    df_ic = df_ic.dropna()  
    return df_ic

df_ic = calculate_factor_ic(df, factor_name)
df_ic.head()
```

```python
def analysis_factor_ic(df_ic):
    ### 初步分析
    print 'Mean of IC: ', df_ic['IC'].mean()
    print 'Median of IC: ', df_ic['IC'].median()
    print 'IC(All, Positive, Negative): ', (len(df_ic), len(df_ic[df_ic.IC>0]), len(df_ic[df_ic.IC<0]))
    
    ### 作图分析
    df_ic['monthly_IC'] = pd.rolling_mean(df_ic.IC, window=21)

    ax = df_ic[df_ic>=0].plot(y=['IC'], style='.r', figsize=(16, 6))
    ax = df_ic[df_ic<=0].plot(y=['IC'], style='.g', ax=ax)
    ax = df_ic.plot(y=['monthly_IC'], style='green', ax=ax)
    ax.legend(['IC(Positve)', 'IC(Negative)', 'IC(Monthly)'], loc='best')
    ax.grid(color='gray', alpha=0.2, axis='y')
    ax.set_axis_bgcolor('white')
    ax.set_ylabel(u'相关系数', fontproperties=font, fontsize=16)
    ax.set_xlabel(u'日期', fontproperties=font, fontsize=16)
    ax.set_title(u"因子和之后10日收益的秩相关系数", fontproperties=font, fontsize=16)

analysis_factor_ic(df_ic)
```

# `因子分组超额收益`

计算因子分组后，10日平均超额收益情况

```python
def quantile_excess_return(df, factor):
    n_quantile = 10    # 统计十分位数
    df_excess_return = pd.DataFrame(index=df.index.unique(), columns=range(1, 11))

    ### 每一天计算一次
    for day in df.index.unique():
        df_one_day = df.ix[day]
        factor_quantile = df_one_day[factor].dropna().quantile(q=np.arange(0.1, 1.1, 0.1)).values
        
        last = 0
        quantile_mean_return = []
        for i in factor_quantile:
            df_piece = df_one_day[(df_one_day[factor] <= i) & (df_one_day[factor] > last)]
            quantile_mean_return.append(df_piece.FR_10.mean())
            last = i
        
        df_excess_return.ix[day] = quantile_mean_return
        
    return df_excess_return
    
df_excess_return = quantile_excess_return(df, factor_name)
df_excess_return.head()
```

```python
def analysis_quantile_excess_return(df_excess_return):
    df_excess_return_mean = pd.DataFrame(df_excess_return.mean(), columns=['excess_return_mean'])
    ax = df_excess_return_mean[df_excess_return_mean>=0].plot(y=['excess_return_mean'], color='r', figsize=(12, 6), kind='bar', width=0.35)
    ax = df_excess_return_mean[df_excess_return_mean<=0].plot(y=['excess_return_mean'], color='g', kind='bar', ax=ax, width=0.35)
    ax.legend(['excess_return_mean(Positve)', 'excess_return_mean(Negative)'], loc='best')
    ax.grid(color='gray', alpha=0.2, axis='y')
    ax.set_axis_bgcolor('white')
    ax.set_ylabel(u'分组超额收益', fontproperties=font, fontsize=16)
    ax.set_xlabel(u'因子十分位组', fontproperties=font, fontsize=16)
    ax.set_title(u"因子选股分组超额收益", fontproperties=font, fontsize=16)
    
    return df_excess_return_mean

df_excess_return_mean = analysis_quantile_excess_return(df_excess_return)
```

# `因子分组反转特征`

因子分组选股的一个月反转分布特征

```python
def quantile_history_return(df, factor):
    n_quantile = 10    # 统计十分位数
    df_history_return = pd.DataFrame(index=df.index.unique(), columns=range(1, 11))

    ### 每一天计算一次
    for day in df.index.unique():
        df_one_day = df.ix[day]
        factor_quantile = df_one_day[factor].dropna().quantile(q=np.arange(0.1, 1.1, 0.1)).values

        last = 0
        quantile_mean_return = []
        for i in factor_quantile:
            df_piece = df_one_day[(df_one_day[factor] <= i) & (df_one_day[factor] > last)]
            quantile_mean_return.append(df_piece.HR_10.mean())
            last = i
        
        df_history_return.ix[day] = quantile_mean_return
        
    return df_history_return
    
df_history_return = quantile_history_return(df, factor_name).dropna()
df_history_return.head(5)
```

# `因子选股 —— 不同五分位数组合回测走势比较`

```python
# 可编辑部分与 strategy 模式一样，其余部分按本例代码编写即可

# -----------回测参数部分开始，可编辑------------
start = '2016-01-01'                       # 回测起始时间
end = '2016-12-31'                         # 回测结束时间
benchmark = 'ZZ500'                        # 策略参考标准
universe = set_universe('A')           # 证券池，支持股票和基金
capital_base = 10000000                     # 起始资金
freq = 'd'                                 # 策略类型，'d'表示日间策略使用日线回测
refresh_rate = 5                           # 调仓频率，表示执行handle_data的时间间隔

df_factor = pd.read_csv('Factor_Analysis_{}.csv'.format(factor_name))     # 读取因子数据
df_factor = df_factor[df_factor.columns[:]].set_index('tradeDate')
factor_dates = df_factor.index.values

quantile_five = 1                           # 选取股票的因子十分位数，1表示选取股票池中因子最小的10%的股票
commission = Commission(0.0005,0.0005)     # 交易费率设为双边千分之一
# ---------------回测参数部分结束----------------


# 把回测参数封装到 SimulationParameters 中，供 quick_backtest 使用
sim_params = quartz.SimulationParameters(start, end, benchmark, universe, capital_base)
# 获取回测行情数据
idxmap, data = quartz.get_daily_data(sim_params)
# 运行结果
results = {}

# 调整参数(选取股票的Q因子五分位数)，进行快速回测
for quantile_five in range(1, 6):
    
    # ---------------策略逻辑部分----------------
    refresh_rate = 5
    commission = Commission(0.0005,0.0005)
    
    def initialize(account):                   # 初始化虚拟账户状态
        pass

    def handle_data(account):                  # 每个交易日的买入卖出指令
        pre_date = account.previous_date.strftime("%Y-%m-%d")
        if pre_date not in factor_dates:            # 因子只在每个月底计算，所以调仓也在每月最后一个交易日进行
            return

        # 拿取调仓日前一个交易日的因子，并按照相应十分位选择股票
        q = df_factor.ix[pre_date][['secID', factor_name]].dropna()
        q_min = q[factor_name].quantile((quantile_five - 1) * 0.1)
        q_max = q[factor_name].quantile(quantile_five * 0.1)
        my_univ = q[(q[factor_name] >= q_min) & (q[factor_name] < q_max)]

        # 调仓逻辑
        univ = [x for x in my_univ.secID.values if x in account.universe]

        if len(univ) < 10:
            return
        
        # 不在目标股票池中的，卖出
        for s in account.valid_secpos:
            if s not in univ:
                order_to(s, 0)

        # 在目标股票池中的，等权买入
        buylist = {}
        v = account.referencePortfolioValue * 1.05 / len(univ) 
        for s in univ:
            buylist[s] = v / account.reference_price[s] - account.valid_secpos.get(s, 0)

        for s in sorted(buylist, key=buylist.get):
            order(s, buylist[s])
    # ---------------策略逻辑部分结束----------------

    # 把回测逻辑封装到 TradingStrategy 中，供 quick_backtest 使用
    strategy = quartz.TradingStrategy(initialize, handle_data)
    # 回测部分
    bt, acct = quartz.quick_backtest(sim_params, strategy, idxmap, data, refresh_rate=refresh_rate, commission=commission)

    # 对于回测的结果，可以通过 perf_parse 函数计算风险指标
    perf = quartz.perf_parse(bt, acct)

    # 保存运行结果
    tmp = {}
    tmp['bt'] = bt
    tmp['annualized_return'] = perf['annualized_return']
    tmp['volatility'] = perf['volatility']
    tmp['max_drawdown'] = perf['max_drawdown']
    tmp['alpha'] = perf['alpha']
    tmp['beta'] = perf['beta']
    tmp['sharpe'] = perf['sharpe']
    tmp['information_ratio'] = perf['information_ratio']
    
    results[quantile_five] = tmp
    print str(quantile_five),
print 'done'
```

```python
fig = plt.figure(figsize=(10, 8))
fig.set_tight_layout(True)
ax1 = fig.add_subplot(211)
ax2 = fig.add_subplot(212)
ax1.grid()
ax2.grid()

for qt in results:
    bt = results[qt]['bt']

    data = bt[[u'tradeDate',u'portfolio_value',u'benchmark_return']]
    data['portfolio_return'] = data.portfolio_value/data.portfolio_value.shift(1) - 1.0   # 总头寸每日回报率
    data['portfolio_return'].ix[0] = data['portfolio_value'].ix[0] / 10000000.0 - 1.0
    data['excess_return'] = data.portfolio_return - data.benchmark_return                 # 总头寸每日超额回报率
    data['excess'] = data.excess_return + 1.0
    data['excess'] = data.excess.cumprod()                # 总头寸对冲指数后的净值序列
    data['portfolio'] = data.portfolio_return + 1.0     
    data['portfolio'] = data.portfolio.cumprod()          # 总头寸不对冲时的净值序列
    data['benchmark'] = data.benchmark_return + 1.0
    data['benchmark'] = data.benchmark.cumprod()          # benchmark的净值序列
    results[qt]['hedged_max_drawdown'] = max([1 - v/max(1, max(data['excess'][:i+1])) for i,v in enumerate(data['excess'])])  # 对冲后净值最大回撤
    results[qt]['hedged_volatility'] = np.std(data['excess_return'])*np.sqrt(252)
    results[qt]['hedged_annualized_return'] = (data['excess'].values[-1])**(252.0/len(data['excess'])) - 1.0
    ax1.plot(data['tradeDate'], data[['portfolio']], label=str(qt))
    ax2.plot(data['tradeDate'], data[['excess']], label=str(qt))


ax1.legend(loc=0, fontsize=12)
ax2.legend(loc=0, fontsize=12)
ax1.set_ylabel(u"净值", fontproperties=font, fontsize=16)
ax2.set_ylabel(u"对冲净值", fontproperties=font, fontsize=16)
ax1.set_title(u"{} - 不同五分位数分组选股净值走势".format(factor_name), fontproperties=font, fontsize=16)
ax2.set_title(u"{} - 不同五分位数分组选股对冲中证500指数后净值走势".format(factor_name), fontproperties=font, fontsize=16)
```
