# 机器学习之算法评估矩阵

```python
import numpy as np
import pandas as pd
import sklearn.model_selection as sm
import sklearn.linear_model as sl
```

对于机器学习，算法评估是非常重要的。选择能够展示机器学习算法模型的准确度的评估矩阵，是计算和比较算法模型最好的方式。算法评估矩阵根据研究目的可分为分类矩阵和回归矩阵。本文介绍的时候分类算法矩阵以逻辑回归为例，回归算法以线性回归为例，使用实际应用中最常用的10折交叉验证来分离术久，并计算展示算法的评估矩阵。

## 1、分类算法矩阵

分类问题是最常见的机器学习问题，一般常用的评估矩阵有：分类准确度、对数损失函数(Logloss)、AUC图、混淆矩阵和分类报告(Classification Report)。

## 1.1 分类准确度

分类准确度就是算法自动分类正确的样本处以所有的样本数的到的结果。通常，准确度越高，分类器越好。这是分类算法中最常见，也是最易被误用的评估参数。准确度确实是一个很好、很直观的评价指标，但是有时候准确度高并不意味算法就一定好。比如说，对某个地区某天地震的预测，假设有一堆的特征作为地震行为的属性，类别却只有两个（0:发生地震；1:不发生地震）。一个不加思考的分类器对每一个测试用例都将类别划分为0，那它就可以达到99%的准确度，但真的地震时，这个歌分类器却毫无察觉，折个分类器造成的损失是巨大的。什么拥有99%准确度的分类器却不是我们想要的，因为数据分布不均衡，类别1的数据太少，完全错分类别1依然可以达到很高的准确度，却忽视了需要关注的事实和现象。代码实现如下。

```python
filename = "iris.data.csv"
names = ["separ-length","separ-width","petal-length","petal-width","class"]
data = pd.read_csv(filename,names=names)
#将数据分为输入和输出数据
s = data.values
x = s[:,0:4]
y = s[:,4]
num_folds = 10
seed = 7
kfold = sm.KFold(n_splits=num_folds,random_state=seed)
model = sl.LogisticRegression()
result = sm.cross_val_score(model,x,y,cv=kfold)
print("算法评估结果准确度：%.3f (%.3f) " % (result.mean(),result.std()))
```

## 1.2 对数损失函数

对数损失, 即对数似然损失(Log-likelihood Loss), 也称逻辑斯谛回归损失(Logistic Loss)或交叉熵损失(cross-entropy Loss), 是在概率估计上定义的。它常用于(multi-nominal, 多项)逻辑斯谛回归和神经网络，以及一些期望极大算法的变体。可用于评估分类器的概率输出。在逻辑回归的推导中，它假设样本服从伯努利分布（0，1分布），然后求满足该分布的似然函数，再去对数、求极值等。而逻辑回归并没有求似然函数的极值，而是把极值最大化当成一种思想，进而推导出它的经验风险函数为：最小负的似然函数[max F(y,f(x)) ➡️ min F(y,f(x)) ]。从损失函数的视角来看，它成了对数的损失函数。对数损失函数越小，模型就越好，对于完美的分类器, 对数损失为 0 ，而且使损失函数尽量使一个凸函数，便于收敛计算。代码实现如下。

```python
filename = "pima_data.csv"
names = ["preg","plas","pres","skin","test","mass","pedi","age","class"]
data = pd.read_csv(filename,names=names)
#将数据分为输入和输出数据
s = data.values
x = s[:,0:8]
y = s[:,8]
num_folds = 10
seed = 7
kfold = sm.KFold(n_splits=num_folds,random_state=seed)
model = sl.LogisticRegression()
scoring = "neg_log_loss"
result = sm.cross_val_score(model,x,y,cv=kfold,scoring=scoring)
print("Logloss %.3f (%.3f) " % (result.mean(),result.std()))
```

## 1.3 AUC图

ROC和AUC是评价分类器的指标。ROC是受试者工作特征曲线的简写，又称感受性曲线。得此名的原因在于曲线上个点反映相同的感受性，它们都是对同一信号刺激的反应，只不过是在几种不同的判定标准下所得结果而已。ROC是反映敏感性和特异性连续变量的综合指标，用构图法揭示敏感性和特异性的相互关系，通过将连续变量设定出多个不同的临界值计算出一系列敏感性和特异性，在以敏感性为纵坐标，1-特异性为横坐标绘制成曲线。AUC是ROC曲线下的面积的简称，顾名思义，AUC的值就是出于ROc曲线下方的那部分面积的大小。通常，AUC的值介于0.5到1.0之间，AUC的值越大，诊断准确率就越高。在ROC曲线上，靠近坐标图左上方的点位敏感性和特异性均较高的临界值。

基于该列联表，定义敏感性指标为：sensitivity=FP/(TP+FN)。敏感性指标又称真正类率，计算的是分类器所识别出的正实例占所有实例的比率。另外，定义负正类率的计算公式为：FPR=FP/(FP+TN)，负正类率计算的是分类器所识别出的负实例占所有实例的比率。定义特异性指标为：Specificity=TN/(FP+TN)=1-FPR。特异性指标又称为真负类率。代码实现如下。

```python
filename = "pima_data.csv"
names = ["preg","plas","pres","skin","test","mass","pedi","age","class"]
data = pd.read_csv(filename,names=names)
#将数据分为输入和输出数据
s = data.values
x = s[:,0:8]
y = s[:,8]
num_folds = 10
seed = 7
kfold = sm.KFold(n_splits=num_folds,random_state=seed)
model = sl.LogisticRegression()
scoring = "roc_auc"
result = sm.cross_val_score(model,x,y,cv=kfold,scoring=scoring)
print("Logloss %.3f (%.3f) " % (result.mean(),result.std()))
```

## 1.4 混淆矩阵

混淆矩阵主要用于比较分类结果和实际测得值，可以把分类结果的精度显示在一个混淆矩阵里面。混淆矩阵是可视化工具，特别适用于监督学习，在无监督学习时一般叫匹配矩阵。混淆矩阵的每列代表预测类别，每列的总数表示预测为该类别的数据的数目；每行代表数据的真实归属类别，每行的数据总数表示该类别的数据的数目。每列中的数值表示真实数据被预测为该类的数目。代码实现如下。

```python
from sklearn.metrics import confusion_matrix
filename = "pima_data.csv"
names = ["preg","plas","pres","skin","test","mass","pedi","age","class"]
data = pd.read_csv(filename,names=names)
#将数据分为输入和输出数据
s = data.values
x = s[:,0:8]
y = s[:,8]
test_size = 0.33
seed = 4
X_train,X_test,Y_train,Y_test = sm.train_test_split(x,y,test_size=test_size,random_state=seed)
model = sl.LogisticRegression()
model.fit(X_train,Y_train)
predicted = model.predict(X_test)
matrix = confusion_matrix(Y_test,predicted)
classes = ["0","1"]
dataframe = pd.DataFrame(data=matrix,index=classes,columns = classes)
print(dataframe)
```

## 1.5 分类报告

在scikit-learn中提供了一个非常方面的工具，可以给出对分类问题的评估报告，Classification_report()方法能够给出精确率(precision)、召回率(recall)、F1值(F1-score)、和样本数量(support)。在这里简单地介绍一下三个数据指标：精确率、召回率和F1值。我们结合讲AUC概念时候提出的一张表格来讲。
* 精确率的公式P=TP/(TP+FP)。计算的是所有被检索到的项目中应该被检索到的项目占的比例。
* 召回率的公式R=TP/(TP+FN)。计算的是所有检索到的项目占所有应该检索到的项目的比例。
* F1值就是精确率和召回率的调和平均，也就是2F1=P+R。

代码实现如下。

```python
from sklearn.metrics import classification_report
filename = "pima_data.csv"
names = ["preg","plas","pres","skin","test","mass","pedi","age","class"]
data = pd.read_csv(filename,names=names)
#将数据分为输入和输出数据
s = data.values
x = s[:,0:8]
y = s[:,8]
test_size = 0.33
seed = 4
X_train,X_test,Y_train,Y_test = sm.train_test_split(x,y,test_size=test_size,random_state=seed)
model = sl.LogisticRegression()
model.fit(X_train,Y_train)
predicted = model.predict(X_test)
report = classification_report(Y_test,predicted)
print(report)
```

## 2、回归算法矩阵

接下来介绍三种评估机器学习的回归算法的评估矩阵：平均绝对误差（Mean Absolute error,MAE）、均方误差（Mean Square Error, MSE）和决定系数（R^2）。

## 2.1 平均绝对误差

平均绝对误差是所有单个观测值与算术平均值的偏差的绝对值的平均值。与平均误差相比，平均绝对误差由于离差被绝对值化，不会出现正负相抵消的情况，因而，平均绝对误差能更好地反映预测值误差的实际情况。代码实现如下。

```python
filename = "housing.csv"
names = ["CRIM","ZN","INDUS","CHAS","NOX","RM","AGE","DIS","RAD","TAX","PRTATIO","B","LSTAT","MEDV"]
data = pd.read_csv(filename,names=names)
#将数据分为输入和输出数据
s = data.values
x = s[:,0:13]
y = s[:,13]
n_splits = 10
seed = 7
kfold =  sm.KFold(n_splits=n_splits,random_state=seed)
model = sl.LinearRegression()
scoring = 'neg_mean_absolute_error'
result = sm.cross_val_score(model,x,y,cv=kfold,scoring=scoring)
print("MAE: %.3f (%.3f) " % (result.mean(),result.std()))
```

## 2.2 均方误差

均方误差是衡量平均误差的方法，可以评价数据的变化程度。均方根误差是均方误差的算术平方根。均方误差的值越小，说明用该预测模型描述实验数据的准确度越高。代码实现如下。

```python
filename = "housing.csv"
names = ["CRIM","ZN","INDUS","CHAS","NOX","RM","AGE","DIS","RAD","TAX","PRTATIO","B","LSTAT","MEDV"]
data = pd.read_csv(filename,names=names)
#将数据分为输入和输出数据
s = data.values
x = s[:,0:13]
y = s[:,13]
n_splits = 10
seed = 7
kfold =  sm.KFold(n_splits=n_splits,random_state=seed)
model = sl.LinearRegression()
scoring = 'neg_mean_squared_error'
result = sm.cross_val_score(model,x,y,cv=kfold,scoring=scoring)
print("MSE: %.3f (%.3f) " % (result.mean(),result.std()))
```

##  2.3 决定系数

决定系数反映因变量的全部变异能通过回归关系被自变量解释的比例。拟合优度越大，自变量对因变量的解释程度就越高，自变量引起的变动占总变动对的百分比越高，观察点在回归直线附近越密集。如$R^{2}$为0.8，则表示回归关系可以解释因变量80%的变异。决定系数的特点：
* 可决系数是非负的统计量
* 可决系数的取值范围：$0\leq R^2\leq1$
* 可决系数是样本观测值的函数，是因随机抽样而变动的随机变量。为此，对可决系数的统计的可靠性也应进行检验。

代码实现如下。

```python
filename = "housing.csv"
names = ["CRIM","ZN","INDUS","CHAS","NOX","RM","AGE","DIS","RAD","TAX","PRTATIO","B","LSTAT","MEDV"]
data = pd.read_csv(filename,names=names)
#将数据分为输入和输出数据
s = data.values
x = s[:,0:13]
y = s[:,13]
n_splits = 10
seed = 7
kfold =  sm.KFold(n_splits=n_splits,random_state=seed)
model = sl.LinearRegression()
scoring = 'r2'
result = sm.cross_val_score(model,x,y,cv=kfold,scoring=scoring)
print("R2: %.3f (%.3f) " % (result.mean(),result.std()))
```
