# WeChat-VaR在险价值

```python
#整理收盘价数据
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
listField = [u'tradeDate', u'closePrice']
etfData = DataAPI.MktFunddAdjBfGet(secID=u"",ticker=u"510050",beginDate=u"",\
                                   endDate=u"",field=listField,pandas="1")
#计算对数收益率序列
etfData['ln_return'] = None
etfData['ln_return'] = (np.log(etfData['closePrice']).diff())*100
etfData.dropna(axis = 0)
etfData.head()
```

```python
#画出对数收益率序列
date = pd.to_datetime(etfData['tradeDate'][1:])
ln_return = etfData['ln_return'][1:]
plt.plot(date, ln_return)
```

```python
#拟合一个正态分布
#示例：拟合一日涨跌幅
mu = np.mean(ln_return)
sigma = np.std(ln_return)
count, bins, ignored = plt.hist(ln_return, 50, normed=True)
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) *
               np.exp( - (bins - mu)**2 / (2 * sigma**2) ),
         linewidth=2, color='r')
plt.axis([-10,10,0,0.4])
plt.show()

print u'均值和标准差分别为：', mu, sigma
print '%5分位点为：', 1.65 * sigma
```

# 风险度量制估计VaR

```python
#计算持有价值10000元的上证50ETF基金的VaR
#前面收益率乘以了100，这里需要除以100
print '持有1天价值10000元的上证50ETF基金的VaR为：', 10000 * 1.65 * sigma /100
print '持有10天价值10000元的上证50ETF基金的VaR为：', 10000 * 1.65 * np.sqrt(10) * sigma/100
```

```python
#非0均值假设
#计算持有价值10000元的上证50ETF基金的VaR
print '持有1天价值10000元的上证50ETF基金的VaR为：', 10000 * (mu + 1.65 * sigma)/100
print '持有10天价值10000元的上证50ETF基金的VaR为：', 10000 * np.sqrt(10) * (np.sqrt(10)*mu + 1.65 * sigma)/100
```

# 计量方法估计VaR

一、ARCH效应检验

```python
#1、通过收益率序列的平方和绝对值检验，
#发现收益率序列平方和绝对值存在明显的聚集和起伏，因此存在ARCH效应（条件异方差性）
plt.figure(figsize = (9, 6))
plt.subplot(211)
plt.plot(ln_return ** 2)
plt.title('')
plt.ylabel('Squared Daily Return')
plt.subplot(212)
plt.plot(np.abs(ln_return))
plt.title('')
plt.ylabel('Absolute Daily Return')
```

```python
#我们直接用收益率序列的各期值减去其均值，得到均值方程噪声项，再对噪声项进行ARCH效应检验
#2、resid^2的pacf图形检验
#发现有严重的自相关性
from statsmodels.graphics.tsaplots import *
resid = ln_return - np.mean(ln_return)
plt.figure(figsize = (9, 4))
ax1 = plt.subplot(111)
plot1 = plot_pacf(resid ** 2, lags=100, ax = ax1)
plt.ylim(-0.1, 0.3)
plt.title('')
plt.ylabel('PACF')
```

```python
#3、resid^2的自相关性检验
#结论：有严重的自相关性
import statsmodels.api as sm
(r,q,p) = sm.tsa.acf(resid ** 2, qstat=True)
data = np.c_[range(1,41), r[1:], q, p]
table = pd.DataFrame(data, columns=['lag', "ACF", "Q", "Prob(>Q)"])
print(table.set_index('lag')) 
```

二、拟合一个AR（4）模型


```python
#根据公众号之前的文章，我们选择AR模型的滞后项为4
ln_return0 = ln_return
ln_return0.index = date
ar = sm.tsa.ARIMA(ln_return0, (4,0,0)).fit()
#ar.summary()
#进行1步和10步均值预测
mean_fore1 = float(ar.forecast(1)[0])
mean_fore10 = float(ar.forecast(10)[0][9])
print u'1日和10日均值预测分别为：', mean_fore1, mean_fore10
```

三、拟合一个AR-GARCH模型

```python
from arch.univariate import ARX
from arch.univariate import ARCH, GARCH
from arch import arch_model
ar = ARX(ln_return, lags = [1, 2, 3, 4])
ar_resid = ar.fit().resid
#AR-GARCH模型建立                
ar.volatility = GARCH(p=1, q=1)
ar_garch = ar.fit()
ar_garch.summary()
```

```python
#画出AR-GARCH波动率
ar_garch.plot(annualize='D')
```

四、进行波动率预测

```python
#conditional volatility = square root of conditional variance
variance_fore1 = ar_garch.forecast(horizon=1).iloc[-1:]['h.1'].values[0]
variance_fore10 = ar_garch.forecast(horizon=10).iloc[-1:]['h.10'].values[0]
volatility_fore1 = np.sqrt(variance_fore1) * np.sqrt(245)
volatility_fore10 = np.sqrt(variance_fore10) * np.sqrt(245)
print u'1日和10日波动率预测分别为：', volatility_fore1, volatility_fore10
```

五、计算多期VaR

```python
print u'持有1天价值10000元的50ETF的VaR为：', \
(mean_fore1 + 1.65 * volatility_fore1) * 10000 /100
print u'持有10天价值10000元的50ETF的VaR为：', \
(mean_fore10 + 1.65 * volatility_fore10) * 10000 /100
```
