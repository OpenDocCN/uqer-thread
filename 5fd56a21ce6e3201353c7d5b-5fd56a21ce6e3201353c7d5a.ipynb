```python
def get_historical_closeIndex(ticker, start, end):
    # get the data for the tickers.  This will be a panel
    p = DataAPI.MktIdxdGet(indexID=ticker,beginDate=start,endDate=end,field=u"tradeDate,indexID,closeIndex",pandas="1")
    p['tradeDate'] = pd.to_datetime(p['tradeDate']) #因为DataAPI所有日期都是以字符串形式存储的，因此此处需要转换成日期时间型,这里很重要！
    # rename the columns
    p.rename(columns={'tradeDate': 'Date', 
                      'closeIndex': 'Close'}, inplace=True)
    # pivot each ticker to a column
    pivoted = p.pivot(index='Date', columns='indexID')
    # and drop the one level on the columns
    pivoted.columns = pivoted.columns.droplevel(0)
    return pivoted
```

```python
closeIndex = get_historical_closeIndex(universe_Idx, start, end)
closeIndex.head()
```

# 历史总是相似?

```python
def zscore(x):
    mu = np.nanmean(x)
    sigma =np.nanstd(x)
    x = (x - mu) / sigma;  
    return x
```

```python
import seaborn
import matplotlib.pyplot as plt

index = '000001'
data = pd.DataFrame()
#第一轮是1991/1-1993/2，指数从128点涨至1558点，随后在1993/2-1994/7进入熊市，上证综指从1500点跌至334点，指数跌幅78%。1994/10-1996/1月上证综指在512点-773点持续了15个月窄幅波动。

data1 = DataAPI.MktIdxdGet(ticker = index, beginDate='19910101', endDate='19960101')
# rename the columns
data1.rename(columns={'closeIndex': '1991-1996'}, inplace=True)
data = data1['1991-1996']

#第二轮是1996/1-2001/6（中间包含了1997/5-1999/5持续24个月的高位震荡），2001年6月12日《减持国有股筹集社会保障资金管理暂行办法》正式发布，在国有股减持的利空压力下市场进入熊市，2001/6-2002/1上证综指从2245点跌至1339点，跌幅40%。熊市之后，2002/1至2005/3上证综指持续38个月的区间波动，2005/3-2005/6市场从1328点跌至998点，市场尾部杀跌源于市场担忧股权分置改革后全流通的冲击。 
data2 = DataAPI.MktIdxdGet(ticker = index, beginDate='19960101', endDate='20050601')
# rename the columns
data2.rename(columns={'closeIndex': '1996-2005'}, inplace=True)
data = pd.concat([data, data2['1996-2005']], axis=1)


#第三轮是2005/6-2014/3/12，2007/10-2008/10上证综指从6124点跌至1664点，跌幅达71%，这次市场暴跌之后没有经历底部的震荡阶段，而是直接从08年底开始快速上涨。2008/10-2009/8，指数10个月涨幅85%。2009/8之后，宏观政策从宽松转向中性，经济增长持续回升，市场在2009/8-2011/4期间高位震荡。2011/4之后，货币政策进一步收紧，伴随经济下行，在2011/4-2012/1期间上证指数从3067点跌至2132点，跌幅30%。熊市之后，在2012/1-2014/7期间指数持续30个月在1849-2444点上下波动。 

data3 = DataAPI.MktIdxdGet(ticker = index, beginDate='20050601', endDate='20140312')
# rename the columns
data3.rename(columns={'closeIndex': '2005-2014'}, inplace=True)
data = pd.concat([data, data3['2005-2014']], axis=1)


#第四轮从2014/3/12-2015/6已经过去，经历了2015/6-2016/1期间的三轮大跌之后，我们判断熊市的主跌浪已经过去，市场将进入震荡阶段。
trade_days = report.trade_pre_n_days()
end = dt.datetime.strptime(trade_days[-1], '%Y-%m-%d')
data4 = DataAPI.MktIdxdGet(ticker =index, beginDate='20140312', endDate=end)
# rename the columns
data4.rename(columns={'closeIndex': '2014-2023'}, inplace=True)
data = pd.concat([data, data4['2014-2023']], axis=1)

data =  zscore(data)
data.plot(figsize=(16,9), grid = False)
plt.legend([u'第一轮（1991-1996）',u'第二轮（1996-2005）',u'第三轮（2005-2014）',u'第四轮（2014-2023）'], prop = font, loc = 'best')
seaborn.despine()
#从A股历史表现来看，除2007/10-2008/10这次熊市以外，其他三次熊市之后市场都会经历一个漫长的震荡筑底过程（15个月-30个月），市场需要经过充分的休整蓄势之后才能够重新迎来新一轮牛市。
```

# 月历效应

```python
def get_return(ticker, start=None, end=None):
    """
    Retrieves prices for ticker from Uqer and computes returns
    based on adjusted closing prices.

    Parameters
    ----------
    ticker : str
        ticker, e.g. "000001.XSHE","000300.ZICN"
    start : '20140101'
        Start date of time period to retrieve
    end : '20160701'
        End date of time period to retrieve

    Returns
    -------
    pandas.DataFrame
        Returns of ticker in requested period.
    """
    if start is None:
        start = '20070101'
    if end is None:
        end = datetime.date.today().strftime('%Y%m%d')
    # 000001.ZICN 上证综指；399001.ZICN 深证成指；399005.ZICN	 中小板指；399006.ZICN 创业板指；
    # 000016.ZICN 上证50；000300.ZICN 沪深300；399905.ZICN 中证500
    if ticker in ['000001.ZICN','399001.ZICN','399005.ZICN','399006.ZICN','000016.ZICN','000300.ZICN','399905.ZICN']: 
        px = DataAPI.MktIdxdGet(indexID=ticker,beginDate=start,endDate=end,field=u"tradeDate,closeIndex",pandas="1")
        px = px.rename(columns={'tradeDate':'Date','closeIndex':'Close'})
    else:
        px = DataAPI.MktEqudAdjGet(secID=ticker,beginDate=start,endDate=end,field=u"tradeDate,closePrice",pandas="1") 
        px = px.rename(columns={'tradeDate':'Date','closePrice':'Close'})
    px['Date'] =  pd.to_datetime(px['Date']) #因为DataAPI所有日期都是以字符串形式存储的，因此此处需要转换
    px = px.set_index("Date")
    px['return'] = px['Close'].pct_change().fillna(0.)
    
    return px['return']
```
