# MACD 技术分析第一指标 V1.0

MACD（Moving Average Convergence & Divergence)堪称技术分析第一指标了，打开一个看盘软件，基本上第一个技术分析指标就是它。  
MACD是利用收盘价的短期（常用为12日）指数移动平均线与长期（常用为26日）指数移动平均线之间的聚合与分离状况，对买进、卖出时机作出研判的技术指标。  
MACD的计算还是比较复杂的，一条移动平均线不难，变成一条指数移动平均线进行平滑计算也不难，变成两条指数移动平均线也还行，但依旧不过瘾，还要把差值继续平滑，再画成线，再以0轴为划分，画成上下的棒，想想也是醉了！  

公式算法:  
短期EMA： 短期（SHORT=12）的收盘价指数移动平均值  
长期EMA： 长期（LONG=26）的收盘价指数移动平均值  
DIF线：　（Difference）短期EMA和长期EMA的离差值  
DEA线：　（Difference Exponential Average）DIF线的（M=9）指数平滑移动平均线  
MACD线：　DIF - DEA的差值（通达信是2倍的DIF-DEA，只关心正负，所以不影响）  

指数加权平滑系数为：  
短期EMA平滑系数： 2/(SHORT+1)  
长期EMA平滑系数： 2/(LONG+1)  
DEA线平滑系数： 2/(M+1)  

MACD买卖策略：
MACD从负到正买进；  
MACD从正到负卖出。  

如果仅仅是这样的标准打法，效果是很差的。所以做一个优化，在DIF>0的前提下买入，过滤掉一部分反弹。  
其他MACD的分析技巧暂时没有涉及，例如背离，位阶搭配，均线搭配等等  
发觉目前这个策略在指数部分应用还不错，但在个股上效果还是不行，其实MACD是个相当不错的指标了（譬如稳定性，也没有KD的钝化问题），要继续优化，使其无愧于第一指标的称号。
