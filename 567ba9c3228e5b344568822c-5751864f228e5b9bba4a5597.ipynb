### 2、值得参考的历史定律
        前一组帖子让我们相信了技术指标冥冥之中自有其独到的价值，但我不相信市场的未来和过去是服从同一分布的，金融市场变幻莫测，未来未必就和过去一样。但请细想，金融市场的参与主体————人类，自身的喜怒哀乐、贪婪或恐惧都是客观存在的，用马克思的话来讲这些是规律性的东西，不以人的意志而转移，所以本文的标题用的是历史定律（不是规律），说的就是因为人本性客观存在的规律而造成的金融市场在某些方面有一定的定律。
        本节有两篇帖子，都是在经历了市场大跌之后的深刻反思总结，别说我没提醒你，真的可以指导散户实盘抄底的！！！

### [`【实盘感悟】遇上暴跌我该怎么做？`](https://uqer.datayes.com/community/share/565d47e3f9f06c6c8a91af49)

- **内容摘要**：作者统计了上证综指在历史上的周度跌幅，发现周度跌幅都低于15%；随后统计了上证综指连续5天的跌幅，以及指数累计下跌10%、15%、20%之后，在随后几天指数的反弹情况，分别从反弹幅度和反弹概率上进行了统计；结果发现，在大盘累计下跌超过10%之后再抄底，在未来5天内都能实现正的盈利，而且下跌幅度越高之后再抄底，获得的收益越高。

- **笔者解读**：本贴是作者实盘经历暴跌之后痛定思痛的总结，全篇没有高深的数学，精华在于统计的视角以及对市场人性的挖掘（这一点和前面历史规律的指导理念是不谋而合的）；市场规律很好总结，但实际行动却有难度，本贴讲出了在暴跌时人性的弱点，极具实盘操作价值！

- **推荐等级**：`★★★★`

### [`2850底是被外资砸出来的？——沪股通数据简单分析`](https://uqer.datayes.com/community/share/564a9754f9f06c4446b48253)

- **内容摘要**：作者从外资流入A股的视角入手，根据通港通成立以来，沪股通每日净流入流出量以及当时上证综指的点位，来近似计算沪股通的历史成本，然后绘制出沪股通历史成本和上证综指的对比图，发现在大盘被砸到3000点以下后，沪股通急速买入A股，形成了现在的2850的底，因为沪股通的历史成本价就在3000点左右，构成了强有力的支持，详见原帖图片。

- **笔者解读**：作者视角新颖，结果非常具有参考价值，可能短线操作有没有太多的实盘参考价值，但对长线投资者绝对是福音，尤其是遇到暴跌之后，知道什么时候该补仓了。

- **推荐等级**：`★★★☆`

### 3、散户的长线量化投资——搬砖配对
        如上两节应该是散户非常喜欢的，也是非常具有实盘的投资价值，更可贵的是它没有资金大小的要求，那我们开始把阶段在提升一下，由散户到机构的过渡阶段————懂长线投资的散户。上面的两节应该都是短线抄底或者等行情来了干一把的散户思维，但有些进阶的散户开始懂得了长线投资，比如有价值型散户就长期持有（至少半年吧）业绩好、有长期稳定分红的蓝筹股，比如中国平安、万科这类，这一类散户应该对本篇会非常感兴趣。
        本篇在长线的投资基础上引入业界经典的配对交易思路，不吃一时大肉，长期吃小肉就够了！

### [`相似公司股票搬砖`](https://uqer.datayes.com/community/share/5555cbfaf9f06c6c7404f8ae)

- **内容摘要**：作者借鉴配对交易的思想，在两个能源电力公司之间搬砖：**川投能源、国投电力**；两家公司由于主营业务、资产结构都非常相似，所以两家的股价长期都保持着非常均衡的关系，作者就在A股价格相比B股价格偏低时，买A股，然后在A股价格相比B股价格偏高时，卖A买B，长期在这两只股票之间“搬砖”，结果发现，对比只长期持有某一只股票而言，搬砖效应能够带来更多的超额收益

- **笔者解读**：国外经典的配对交易模型在A股市场是很难实施的，因为缺乏有效地做空个股的方式，但是原作者借鉴配对交易思想，给偏好于长期投资的散户带来福音；其实，配对股不仅只有能源类，在其他行业更是数不胜数（比如银行），假设你是长期投资者，比如想长期投资银行，那么搬砖策略就是你不错的选择，相比你长期持有某一只银行而言，简单的搬砖策略带给你的超额收益绝对是你想象不到的（笔者私底下有测试过，超额收益绝对比想象中的高很多）

- **推荐等级**：`★★★★`
