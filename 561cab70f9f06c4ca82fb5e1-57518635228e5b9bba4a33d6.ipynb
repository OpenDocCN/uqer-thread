# 【 大师系列 】 价值选股法 —— 三一投资管理公司

#  *价值投资策略*

* `三一投资管理公司（Trinity Investment
Management Corporation）`

>   三一投资管理公司(Trinity Investment Management Corporation) 创立于
1974 年，原本只提供投资机构客户投资研究建议，自1980 年开始为客户管理投资
组合，1999 年成为美国最大的共同基金及投资管理集团之一的欧本海默基金公司
(Oppenheimer Funds, Inc.)的一员，至2000 年第一季底为止，三一投资管理公司
为客户管理的资产总额达42 亿美元，其投资哲学以价值导向为主，并认为建立成功
的价值型投资组合，只需要以单纯而简化的概念及系统性的设计即可。”


> > `选股条件` ：


>1. 条件_1   选取本益比最低的前30%公司 （`PE`）
>2. 条件_2   股价账面价值比最低的前30%公司 (`PB`)
>3. 条件_3   股利收益率最高的前30%公司 (`派息水平`)
>4. 条件_4   若选出股票超过30 个，选取股利收益率最高的前30 个进入组合
