如果我的理解没错就是，

$ R_p - R_r = \beta * (R_M - R_r) + \beta_1 * HML + \beta_2 * SMB + \alpha $

这里我懒于计算 $ R_r $ 即，无风险利率，因为如果它是个常数的话，可以看成在$\ \alpha$ 代表不同的含义

HML 对应PB，将PB分成三档。最高的为成长性（H），最低的为价值型（L），高组收益率平均减去低组收益率平均；

SMB 对应LCAP，将LCAP分成三档。最小的为小公司（S），最大的为大公司（B），小公司收益率平均减去大公司收益率平均。

其中，收益率为月度收益。 $ R_p $ 是单只股票的月度收益率. $R_M$是沪深300的月度收益率

然后看到有人用了个策略是，alpha总会归零，于是每月的回归之后挑alpha最负的是个股票……

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from CAL.PyCAL import *    # 导入CAL
import statsmodels.formula.api as smf
import statsmodels.api as sm
```

```python
cal = Calendar('China.SSE')
initialDt = Date(2013,1,31)
while initialDt < Date.parseISO('2016-06-30'):
    monthEndDate = initialDt.strftime('%Y%m%d')
    ed = initialDt.strftime('%Y%m%d')
    bd = Date(initialDt.year(),initialDt.month(),1).strftime('%Y%m%d')
    initialDt = cal.advanceDate(initialDt, '1M', BizDayConvention.Nearest, endOfMonth = True)
    # benchmark = 'hs300'
    df1 = DataAPI.MktEqumAdjGet(secID=u"",ticker=u"",monthEndDate=monthEndDate,isOpen=u"",beginDate=u"",endDate=u"",field=u"secID,chgPct",pandas="1")
    df2 = DataAPI.MktStockFactorsOneDayGet(tradeDate=monthEndDate,secID=u"",ticker=u"",field=u"secID,PB,LCAP,tradeDate",pandas="1")
    df = pd.merge(df1,df2,on='secID')
    hs300 = DataAPI.MktIdxdGet(tradeDate=u"",indexID=u"000300.ZICN",ticker=u"",beginDate=bd,endDate=ed,field=u"",pandas="1")
    # riskless return
    rm = (hs300['closeIndex'].iloc[-1] - hs300['closeIndex'].iloc[0]) / hs300['closeIndex'].iloc[0]
    df['rm'] = rm
    df.dropna(inplace=True)
    df.sort('LCAP',inplace=True)
    # Small minus Big
    length = len(df.index)
    SMB = df[0:int(1.0 / 3*length)]['chgPct'].mean() - df[int(2.0 / 3*length):]['chgPct'].mean()
    df['SMB'] = SMB
    df.sort('PB',inplace=True)
    HML = df[int(2.0 / 3*length):]['chgPct'].mean() - df[0:int(1.0 / 3*length)]['chgPct'].mean()
    df['HML'] = HML
    ttlist.append(df)
res = pd.concat(ttlist,ignore_index=True)
```

```python
res1= res[res['secID'] == '600000.XSHG']
print(res1)
ft = smf.ols('chgPct ~ rm + SMB +HML',res1).fit()
sm.graphics.plot_regress_exog(ft,1)
print(ft.summary())
```
