下面是一个很有趣的回测：

1. 回测时间2011-1-1  ~  2016-6-1
2. 回测的股票池是HS300
3. 策略的主要思想是：  平均买入HS300的所有股票

有意思的结论：  我的年化收益跑赢指数14%！！！！！！！！！！！！！！


这显然是“不合理”的。造成这个“不合理”的原因目前能想到如下：
1. HS300的成分股是静态的， 如果回测期间HS300编制发生改变， 则直接去掉了剔除的股票。 导致回测的股票池本身就是以淘汰劣质股之后的股票来回测。（未来数据）
2. 权重， HS300的每个股票的权重不一样， 回测用的是等权。 （个人觉得等权不会造成14%的年化收益的偏差）
3. 剔除的停牌， ST， 流通市值的影响。 （这个影响也不应该造成这么大的偏差）
4. 股票分红除权对指数的影响。 （每年1%够了吧？）

哪位大神帮忙找找这个回测的BUG， 如果回测的偏差年化14%，  那进行回测几乎就失去了意义！！！！！！！！
