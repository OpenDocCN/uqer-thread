# 基于市场择时轮动的策略系列（非周期A股和创业板）

基于格式成长和迈克尔.喜伟轮动策略（非周期A股和创业板）：
    1.选取A股与创业板指数作为换股参考
    2 比较全A股的20日涨幅与创业板指数的20日涨幅
        2.1 如果全A股的20日涨幅与创业板指数的20日涨幅都小于0则清空所有股票；
        2.2 如果全A股的20日涨幅大于创业板指数的20日涨幅；
            2.2.1 提取非周期性A股的单季度财务指标，包括 EPS 每股收益 niYOY 净利润同比增长率和此财务报表的截止日期 
            2.2.2 根据格氏成长股公式计算企业价值；
            2.2.3 计算价值比，选出价值比最高的10支股票；
            2.2.4 卖出不在买入股票列表的股票，等权买入股票列表中的股票；
        2.3 如果全A股的20日涨幅小于创业板指数的20日涨幅
            2.3.1 筛选利用迈克尔.喜伟的收益型投资法在创业板中选取股票；
            2.3.4 卖出不在买入股票列表的股票，等权买入股票列表中的股票；

基于格式成长的市场轮动策略（非周期A股和创业板）：
    策略调仓频率每周调仓(Weekly(4), ['09:40'])，为了降低策略的最大回撤与进一步提高策略的收益率与夏普比例，本文利用格式成长选股模型,以HS300为股票池。创业板以流通市值'LFLO'，经营活动产生的现金流量净额与企业价值之比 CFO2EV（Cash provided by operations to enterprise value）营业利润增长率 OperatingProfitGrowRate,分析师推荐评级'REC'选取股票。
    1.选取HS300指数与创业板指数作为换股参考
    2 比较HS300指数的20日涨幅与创业板指数的20日涨幅
        2.1 如果HS300指数的20日涨幅与创业板指数的20日涨幅都小于0则清空所有股票；
        2.2 如果HS300的20日涨幅大于创业板指数的20日涨幅；
            2.2.1 提取HS300股票的单季度财务指标，包括 EPS 每股收益 niYOY 净利润同比增长率和此财务报表的截止日期 
            2.2.2 根据格氏成长股公式计算企业价值；
            2.2.3 计算价值比，选出价值比最高的10支股票；
            2.2.4 卖出不在买入股票列表的股票，等权买入股票列表中的股票；
        2.3 如果HS300的20日涨幅小于创业板指数的20日涨幅
            2.3.1 筛选创业板中流通市值最小的66.67%的股票；
            2.3.2 在2.3.1得到的股票中筛选CFO2EV最高的30.00%的股票（相对于所有创业板股票的30.00%）；
            2.3.3 在2.3.2得到的股票中筛选OperatingProfitGrowRate最高的30.00%的股票；
            2.3.4 在2.3.3得到的股票中取得评级最高的10支股票；
            2.3.5 卖出不在买入股票列表的股票，等权买入股票列表中的股票；

上面这个策略创业板选股逻辑：
主要来自研报实验中表现较好的因子，比如流通市值（规模因子）比总市值表现较好，选取流通市值（从小到大排列）的前2/3股票的股票，
经营活动产生的现金流量净额（成长性因子）与企业价值之比，考察企业的经营状况和估值水平。在前期选股基础上选取前30%的股票。
分析师评级（分析师预测因子）表现较好，是一个主观因素因子。是炒股常用的参考标准，在前期基础上选取表现最好的十支股票。
