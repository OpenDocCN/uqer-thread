之前有段时间学习了一点机器学习的知识，其实还是比较容易的，关键是训练数据的特征要明显。

下面是之前随便写的一个随机森林的模型，取前60天的数据，然后1-20,2-21,3-22这样每次取20天来预测之后10天的涨跌。

本帖主要是为了抛砖引玉，因为本人才疏学浅，机器学习方面我觉得我的数学是硬伤，难以有大的突破了。

所以在看到社区里机器学习帖子不多的情况下，为了增加机器学习的讨论，滥竽充数一下。


***  

机器学习，最重要的是特征。

>我做这个模型的想法是：行情总是龙头股带动起来的，一部分股票先涨，带动另一部分股票涨。

>那么学习前者的上涨方式，可以指导预测后涨的股票。

我把*MA5、MA10、MA20以及价格*的比例作为特征，这样就避免了有些价格高，有些价格低的问题，全部采用比例，也就相当于均线和价格之间的位置关系来预测了。

回测过程较为漫长，因此分成了一年一年去做。

惊奇的发现2013年的测试还是比较不错的。如果`set_universe`存在幸存者偏差的话，51.8%的alpha也足够覆盖幸存者偏差的alpha了。

然而，2014之后的效果越来越差了，2015前半年有正的alpha，后半年就负的比较厉害了。

不明白这样的原因是什么，真的是幸存者偏差？
如果是这样，那么自己先选定一些比较好的股票池进行学习应该会有用。

如果选定其他的基本面因子做学习的话，在大熊市来临的时候经常会大雪崩，不知道大神的机器学习策略在2015年下半年的熊市是如何应对的。
