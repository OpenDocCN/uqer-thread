摘   要：价值投资作为一种“抱朴守拙”的投资理念，为当前股票投资提供了最为纯粹的投资范式，广受政学业三界推崇。本文仿照FAMA五因子模型的设计理念，选取盈利因子，账面市值比因子，规模因子作为基本面的初始判断因子，以此构建价值投资量化交易策略，借助2019年至今(中国基金业最为发达的时段)中证800股票数据，回测检验上述因子在量化投资中的作用效果。回测发现，盈利因子，账面市值比因子发挥的作用的方式与五因子一文中的观点有较大的相似之处，而市值因子在回测时段内的表现却出现与FAMA相反的结论。总体而言，上述三类因子均在本次回测实验中对选股策略起到为正的增量贡献，这也在某种程度上与价值投资理念互相验证。
尽管价值投资理念广受政学业三界推崇，然而在我国证券市场上，这一理念的践行者却为数不多。本文认为，除去行业规则等外在因素，投资者对于价值投资的顾虑，更多的表现在“价值真伪”方面.由于市场上的信息充斥着噪音，基于传统的基本面分析在交易过程中的收益很容易受到虚假信息的干扰。同理，在量化投资的应用上，市场虚假信息的干扰同样也是价值投资策略的担忧所在。故此，本文尝试以优矿平台的换手波动率作为优化因子进一步过滤市场虚假信息干扰，回测结果发现，虚假信息的过滤能够显著提升价值投资策略的有效性。
关键词：价值投资；因子选股；因子优化； 

一、	研究设计
（一）	样本选择与数据来源
本文通过优矿平台进行量化交易回测，检验价值投资类因子在投资决策中发挥的作用。将中国2019年——2021年今中证800成分股上市公司作为研究样本，其中因子数据来自优矿DataAPI接口。考虑到分析数据的客观性，本文在初始样本的基础上做了如下处理：1.剔除了每日因IPO,涨跌停，制度管理等停牌休市的股票，消除了现实交易对此类股票存在的交易不确定性。2.剔除了样本期限内存在ST标记以及退市的企业，防止量化交易策略投资风险较高的股票。3.对因子数据采用前置一日的处理，避免存在事后决策的现象。
（二）	因子设定和检验方式
（1）	初始策略因子
本文根据FAMA五因子模型，选择其中与股票价值面相关的因子角度进行因子组合构建，其中包括账面市值比因子，盈利质量因子，规模因子。与FAMA不同的是，本文选取的因子并非以超额收益率作为分解对象，而是将上述类型的因子作为量化交易决策中股票交易买卖的信号。
账面市值比因子：结合当前的证券市场环境，选择使用最为成熟的市盈率(PE)和市净率(PB)作为账面市值比因子。
盈利质量因子：以权益回报率(ROE)度量股票公司的盈利水平。计算公式：权益回报率(ROE) = 净利润(TTM)/股东权益。
规模因子：以对数市值(LCAP)作为企业规模大小的分组方式。计算公司：对数市值(LCAP) = ln(市值)
（2）	检验方式
首先，根据FAMA的研究结论，账面市值比(HML)作为冗余因子，很大程度上被其他因子所解释(其中包括本文的规模因子)，故此，本文将账面市值比因子(PB,PE)作为全局初筛因子，根据市场惯例，规定选股目标必须满足市盈率介于0与100之间，市净率介于0与20之间。
进一步地，参考FAMA的分组方式，对盈利质量因子，规模因子采用三分位数分组，参照FAMA的结论，选择最大分位数的盈利质量因子，最小分位数的规模因子作为选股筛选标准，逐步添加盈利质量因子，规模因子，检验因子策略在量化交易中的回测表现和增量贡献。
（3）	修正因子
成长性因子：由于标的股票池为中证800成分股，大企业在发展至瓶颈期时容易出现增速放缓甚至负增长的情况，因此引入成长性因子，以营业收入增长率(OperatingRevenueGrowRate)度量企业成长性水平。计算公式为：营业收入增长率=(今年营业收入(TTM)/去年营业收入(TTM))-1

二、	回测结果与分析
（一）	基准回测
根据实验设计逐步添加因子进行回测检验，下图报告了添加ROE因子后的回测结果，相对于参照对象的原始股票池而言，ROE因子的表现尚可，信息比率为1.04。在收益指标方面，ROE因子策略年化收益率高达33.1%（高于参照年化收益率25.9%），在策略实施的11个月后开始出现超额收益率；在风险指标方面，贝塔系数小于1，因子所提供的选股策略也表现出优于大盘的表现。

为进一步地为策略加入规模因子作为选股过滤信号，如下图所示，策略年化收益率从原先的33.1%下降至17.1%，说明了小市值这一因子难以在本策略中表现出价值效用，甚至出现了负面的因子价值，对此，我们不妨假设，FAMA的市值因子策略在本次策略以及选股池中结论可能相反，对此，我们反其道而行，选择最大分位数的因子数据作为选股策略的新指标，如下第三段回测图所示。

从下图的结果中发现，随着引入大市值的因子影响因素，策略结果显著提升，可能的原因如下。首先，本次回测选择的股票池为中证800成分股，普遍为规模较大的企业，以此类企业作为研究对象时，小市值这一作用结论难免形成悖论；其次，以中证800作为研究对象时，对于成熟企业群，市值与其自身企业的质量会有更大的相关性；最后，从近两年的市场情绪来看，由于我国公募基金行业发展迅猛，使得市场资金更多的集中在基金机构当中，出于对风控的需求，基金持仓通常更偏向于考虑市值较大，风险较小的成熟企业股票。
尽管引入大市值因子作为选股策略能够显著提升策略的收益水平，但相较于原本的ROE选股策略而言，与风险相关的指标也显著提升（贝塔和回测），本文认为，该因子的引入尚未在中国中证800的股票池环境下，替代小市值因子的增量贡献。

（二）	因子策略修正
细读FAMA三因子模型资料，在选择规模因子时，认为小规模企业具备较大的增长空间，其判断依据更多来自企业的成长属性，而前文选择引入大市值因子时，规模较大的企业也会存在更大的发展瓶颈，但这种猜想依旧需要数据的描述性统计来进一步印证，通过pandas的describe功能获取因子数据的描述性统计结果，如下所示。（这一部分结果是离线完成的，就是用DataAPI获取一下数据本地看一下描述性统计。结果就没放上来了，在自己的word文档里）

上表报告了样本数据的描述性统计结果，其中营业收入增长率因子（OperatingRevenueGrowRate）数据的最小值为-0.9966，且存在310074个样本发为负的数据，而因受增长率因子作为股票分析的成长性因子，出现这样的数据表现也说明了中证800股票池有部分股票在回测期间存在成长水平不佳的状况，这一因子也很有可能会干扰到策略的风险收益状况。
显然，成长因子是值得我们在执行策略时重点关注的因素，故此，本文从成长因子中，选择营业收入增长率因子作为因子优化策略，剔除股票池中成长率为负的股票。最终回测结果如下图所示，引入了成长因子后，能够显著提高决策的收益率，同时也使得策略的风险指标贝塔值有所下降（从原先的1.06下降至1.04）。

展性研究：基于换手波动率因子的策略优化
对于价值投资理念而言，投资者最担心的问题莫过于获取到虚假的基本面信息。通常情况下，较为劣质的企业会通过财务粉饰，谎报利好等方式来包装自身，导致市场信息存在噪声，但这种方式在证券市场的技术面会有所反馈，类似虚假信息从散播到辟谣的过程中，企业股票的换手率会存在较大的波动，基于这一设想，本文尝试引入换手波动率因子Volatility，进一步对上述的价值投资策略进行优化。
在此之前，本文获取了回测期间内换手波动率因子的周信息比率时序图，如下图5所示。从结果可以看出，在单独对其考察该情况下，该因子的信息比率在回测期间内绝大部分时间保持接近0.1的信息比率，数据表现尚可，但依旧无法得出其在结合其他因子选股时的信息比率表现。

```python
#计算IC
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False

# 取沪深300成分股
hs300_frame = DataAPI.IdxConsGet(secID=u"",ticker=u"000906",isNew=u"",intoDate=u"",field=u"",pandas="1")
hs300_frame[hs300_frame.isNew==1]
tickers = list(hs300_frame.consTickerSymbol.values)
factor_frame = DataAPI.MktStockFactorsDateRangeGet(secID=u"",ticker=tickers,beginDate=u"20180101",endDate=u"20210605",
                                                   field=['secID', 'tradeDate', 'Volatility', 'DAVOL10'],pandas="1")
tfactor_frame = factor_frame
#tfactor_frame
tfactor_frame.head()
import pandas as pd
import numpy as np
# 取A股2019-01-01到2021-06-05的每周周末因子值
# 先取交易日历
calendar_frame = DataAPI.TradeCalGet(exchangeCD=u"XSHG,XSHE",beginDate=u"20180101",endDate=u"20210605",field=u"",pandas="1")
# 只取周末日期
calendar_frame = calendar_frame[calendar_frame.isWeekEnd==1]
calendar_frame.head()
factor_frame_list = []
# 遍历每一天，拿到因子数据
for tdate in calendar_frame.calendarDate.values:
    factor_frame = DataAPI.MktStockFactorsOneDayGet(tradeDate=tdate,ticker=tickers,field=["secID",'tradeDate','Volatility', 'DAVOL10'],pandas="1")
    factor_frame_list.append(factor_frame)
# 合并所有天的因子值
tfactor_frame = pd.concat(factor_frame_list)
tfactor_frame['tradeDate'] = tfactor_frame['tradeDate'].apply(lambda x: x.replace("-",""))
tfactor_frame.head()

bak_frame = tfactor_frame.copy()

print u'tfactor_frame 因子数据data:\n'
print tfactor_frame.head().to_html()
# 拿到周收益率数据
weekly_return_frame = DataAPI.MktEquwGet(secID=u"",ticker=tickers,weekEndDate=u"",beginDate=u"20180101",endDate=u"20210605",
                                          isOpen=u"",field=["secID",'endDate','return'],pandas="1")
weekly_return_frame['endDate'] = weekly_return_frame['endDate'].apply(lambda x: x.replace("-",""))
print u'monthly_return_frame 每周收益率data:\n'
print weekly_return_frame.head().to_html()

# 拿到交易日历，用来做对齐
calendar_frame = DataAPI.TradeCalGet(exchangeCD=u"XSHG,XSHE",beginDate=u"20180101",endDate=u"20210605",field=u"",pandas="1")
calendar_frame = calendar_frame[calendar_frame.isWeekEnd==1][['calendarDate', 'isWeekEnd']]
calendar_frame['calendarDate'] = calendar_frame['calendarDate'].apply(lambda x: x.replace("-",""))
calendar_frame['prev_week_end'] = calendar_frame['calendarDate'].shift(1)
calendar_frame = calendar_frame[['calendarDate', 'prev_week_end']]
calendar_frame.rename(columns={"calendarDate":"week_end"}, inplace=True)

print u'calendar_frame 交易日历关系\n'
print calendar_frame.head().to_html()
# 将数据对齐
tfactor_frame = tfactor_frame.merge(calendar_frame, left_on=['tradeDate'], right_on=['prev_week_end'], how='left')
print '对齐交易日历'
print tfactor_frame.head().to_html()

tfactor_frame = tfactor_frame.merge(weekly_return_frame, left_on=['secID', 'week_end'], right_on=['secID', 'endDate'], how='left')
print '对齐下期收益率的dataframe'
print tfactor_frame.head().to_html()
import scipy.stats as st
def calc_normal_ic(df):
    tmp_df = df.copy()
    tmp_df.dropna(inplace=True)
    return tmp_df['Volatility'].corr(tmp_df['return'])

# 用scipy的包计算
def calc_normal_ic_2(df):
    tmp_df = df.copy()
    tmp_df.dropna(inplace=True)
    return st.pearsonr(tmp_df['Volatility'], tmp_df['return'])[0]

# 用scipy的包计算
def calc_rank_ic(df):
    tmp_df = df.copy()
    tmp_df.dropna(inplace=True)
    return st.spearmanr(tmp_df['Volatility'], tmp_df['return'])[0]

# 计算Normal IC
normal_ic_frame = tfactor_frame.groupby(['tradeDate']).apply(calc_normal_ic)
print "Normal IC:"
print normal_ic_frame.head()

# 计算Rank IC
rank_ic_frame = tfactor_frame.groupby(['tradeDate']).apply(calc_rank_ic)
print "-"*20
print "Rank IC:"
print rank_ic_frame.head()
import matplotlib.pyplot as plt
ticker_spacing = 50
fig = plt.figure(figsize=(18, 6))

ax1 = fig.add_subplot(121)

ax1=normal_ic_frame.plot(ax=ax1)
ax1.xaxis.set_major_locator(ticker.MultipleLocator(ticker_spacing))
ax1.set_title("Normal IC")
ax1.grid(True)

ax2 = fig.add_subplot(122)
ax2 = rank_ic_frame.plot(ax=ax2)
ax2.xaxis.set_major_locator(ticker.MultipleLocator(ticker_spacing))
ax2.set_title("Rank IC")
ax2.grid(True)

print "mean normal IC:",normal_ic_frame.mean(), "VS  mean Rank IC:", rank_ic_frame.mean()
```

随即，本文引入换手波动率因子进行分组回测，如下图6所示，从回测结果来看，纯股票多头时，随着分位数的下降，年化收益率呈现逐级递减的趋势，尽管从时走势图中可以看出，这一因子的效用存在约一年的滞后效应，但依旧在某种程度上可以说明该因子具备一定的价值。

```python
#这一段分组检验的框架在新手文章里面有，结合自己的策略小改一下
import time
from CAL.PyCAL import * 

start_time = time.time()
# ------------------------------------------------- 设定回测参数 ------------------------------------------------------

start = '2019-01-01'                       # 回测起始时间
end = '2021-06-04'                         # 回测结束时间
universe = universe = DynamicUniverse('ZZ800')      # 证券池，支持股票、基金、期货、指数四种资产  #后面用Volatility 前20% |Factor.Volatility.pct_range(0, 0.1)


benchmark = '000906.ZICN'                        # 策略参考标准
freq = 'd'                                 # 策略类型，'d'表示日间策略使用日线回测，'m'表示日内策略使用分钟线回测
refresh_rate = Monthly(1)                        # 调仓频率，表示执行handle_data的时间间隔，若freq = 'd'时间间隔的单位为交易日，若freq = 'm'时间间隔为分钟
num = 20                                   #持仓数目  
capital_base = 10000000
accounts = {
    'fantasy_account': AccountConfig(account_type='security', capital_base=10000000)
}

# 调用quick_backtest方法
sim_params = quartz.SimulationParameters(start, end, benchmark, universe, capital_base, refresh_rate=refresh_rate, accounts=accounts)
# 获取回测行情数据
data = quartz.get_backtest_data(sim_params)
# 运行结果
results = {}

# 遍历每一组，进行回测
group_num = 5
loop_num = group_num + 1
cheng = 1.0/group_num
for quantile_five in range(1, loop_num):  
    def initialize(context):                   # 初始化虚拟账户状态
        pass

    def handle_data(context): 
        previous_date = context.previous_date.strftime('%Y-%m-%d')
        account = context.get_account('fantasy_account')
        universe = context.get_universe('stock', exclude_halt=True)
        hist = context.history(symbol=universe, attribute=['ROE','Volatility','LCAP','PE','PB'], 
                               time_range=1, style='tas')[previous_date]
        hist = hist[(hist.PE>0) & (hist.PB>0) &(hist.PE<100) & (hist.PB<20)]
        signal = hist.sort_values(by = 'Volatility',ascending=True)
        hist = signal.iloc[len(signal)*(quantile_five-1)//5:len(signal)*quantile_five//5]
        signal1 = hist.sort_values(by ='ROE',ascending=False)
        signal_length1 = len(signal1)
        signal1 = signal1[:(signal_length1//3)]  
        signal2 = hist.sort_values(by = 'LCAP',ascending=False)
        signal_length2 = len(signal2)
        signal2 = signal2[:(signal_length2//3)]
        #signal = hist.sort_values(by = ['ROE','LCAP'],ascending=[False,True])
        #print(len(signal))
        list1 = list(signal1.index)
        list2 = list(signal2.index)
        target_position = list(set(list1).intersection(set(list2)))  
        
        #获取当前用户信息
        account = context.get_account('fantasy_account') 
        current_position = account.get_positions(exclude_halt=True) 
        
        current_position_all = account.get_positions()
        # 卖出当前持有，但目标持仓没有的部分
        for stock in set(current_position).difference(target_position):
            account.order_to(stock, 0)
            
        # 根据目标持仓权重，逐一委托下单
        for stock in set(target_position).difference(current_position_all):
            account.order_pct(stock, 1.0/len(target_position))

    # 生成策略对象
    strategy = quartz.TradingStrategy(initialize, handle_data)
    
    # 开始回测,
    bt, perf = quartz.quick_backtest(sim_params, strategy, data=data)

    # 保存运行结果，1为因子最强组，5为因子最弱组
    results[loop_num - quantile_five] = {'max_drawdown': perf['max_drawdown'], 'sharpe': perf['sharpe'], 'alpha': perf['alpha'], 'beta': perf['beta'], 'information_ratio': perf['information_ratio'], 'annualized_return': perf['annualized_return'], 'bt': bt}    

    print ('backtesting for group %s..................................' % str(quantile_five))
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
sns.set_style('white')

fig = plt.figure(figsize=(10,8))
fig.set_tight_layout(True)
ax1 = fig.add_subplot(211)
ax2 = fig.add_subplot(212)
ax1.grid()
ax2.grid()

for qt in results:
    bt = results[qt]['bt']

    data = bt[[u'tradeDate',u'portfolio_value',u'benchmark_return']]
    data['portfolio_return'] = data.portfolio_value/data.portfolio_value.shift(1) - 1.0   # 总头寸每日回报率
    data['portfolio_return'].ix[0] = data['portfolio_value'].ix[0]/	10000000.0 - 1.0
    data['excess_return'] = data.portfolio_return - data.benchmark_return                 # 总头寸每日超额回报率
    data['excess'] = data.excess_return + 1.0
    data['excess'] = data.excess.cumprod()                # 总头寸对冲指数后的净值序列
    data['portfolio'] = data.portfolio_return + 1.0     
    data['portfolio'] = data.portfolio.cumprod()          # 总头寸不对冲时的净值序列
    data['benchmark'] = data.benchmark_return + 1.0
    data['benchmark'] = data.benchmark.cumprod()          # benchmark的净值序列
    results[qt]['hedged_max_drawdown'] = max([1 - v/max(1, max(data['excess'][:i+1])) for i,v in enumerate(data['excess'])])  # 对冲后净值最大回撤
    results[qt]['hedged_volatility'] = np.std(data['excess_return'])*np.sqrt(252)
    results[qt]['hedged_annualized_return'] = (data['excess'].values[-1])**(252.0/len(data['excess'])) - 1.0
    ax1.plot(data['tradeDate'], data[['portfolio']], label=str(qt))
    ax2.plot(data['tradeDate'], data[['excess']], label=str(qt))
    

ax1.legend(loc=0)
ax2.legend(loc=0)
ax1.set_ylabel(u"净值", fontproperties=font, fontsize=16)
ax2.set_ylabel(u"对冲净值", fontproperties=font, fontsize=16)
ax1.set_title(u"因子不同五分位数分组选股净值走势", fontproperties=font, fontsize=16)
ax2.set_title(u"因子不同五分位数分组选股对冲后净值走势", fontproperties=font, fontsize=16)

# results 转换为 DataFrame
results_pd = pd.DataFrame(results).T.sort_index()

results_pd = results_pd[[u'alpha', u'beta', u'information_ratio', u'sharpe', u'annualized_return', u'max_drawdown',  
                         u'hedged_annualized_return', u'hedged_max_drawdown', u'hedged_volatility']]

cols = [(u'风险指标', u'Alpha'), (u'风险指标', u'Beta'), (u'风险指标', u'信息比率'), (u'风险指标', u'夏普比率'), (u'纯股票多头时', u'年化收益'),
        (u'纯股票多头时', u'最大回撤'), (u'对冲后', u'年化收益'), (u'对冲后', u'最大回撤'), (u'对冲后', u'收益波动率')]
results_pd.columns = pd.MultiIndex.from_tuples(cols)
results_pd.index.name = u'五分位组别'
results_pd.to_excel('myresult.xlsx')
results_pd
```
