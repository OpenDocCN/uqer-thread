# XAUUSD

以过去十年（2007年至2017年）黄金收盘价作数据，做各类分析，描述其价格波动规律。
这些均作为后续机器学习的更多规则输入。

```python
import pandas as pd
import matplotlib.pyplot as plt

def compute_daily_returns(df):
	daily_returns = (df / df.shift(1)) - 1
	daily_returns.ix[0, :] = 0
	return daily_returns

def plot_data_with_mean_std(df):
    mean = df['Close'].mean()
    print "均值 =", mean
    std = df['Close'].std()
    print "标准差 =", std
    print "峰度 =", df['Close'].kurtosis()
    df.hist(bins=80)
    plt.axvline(mean, color='w', linestyle='dashed', linewidth=2)
    plt.axvline(std, color='r', linestyle='dashed', linewidth=2)
    plt.axvline(-std, color='r', linestyle='dashed', linewidth=2)
    plt.title('XAUUSD Daily Increased')
    plt.show()

xau_location = '10_Y_DXAUUSD.csv'
df_xau = pd.read_csv(xau_location,index_col = "Date", parse_dates=True, usecols=['Date', 'Close'])
plot_data_with_mean_std(compute_daily_returns(df_xau))
```

以上是XAUUSD在过去十年的日线价格涨跌图，由图可以看到日涨幅基本在0.1 和 -0.1 之间，呈正态分布，具体均值和标准差如下： 
均值 = 0.000198546618577（白线标注） 
标准差 = 0.0107013345952 （红线标注） 
说明日涨幅的均值大概是0.0002，一般情况下涨幅大概率会落在两条红线之间，即日涨幅范围 [-0.0105， 0.0109]

黄金和美元的走势一般讲是相反的，但相反的程度如何，相关度有多大，下面做了分析。

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

def compute_daily_returns(df):
	daily_returns = (df / df.shift(1)) - 1
	daily_returns.ix[0, :] = 0
	return daily_returns

def normalize_data(df):
    return df/df.ix[0,:]

xau_location = '10_Y_DXAUUSD.csv'
df_xau = pd.read_csv(xau_location,index_col = "Date", parse_dates=True, usecols=['Date', 'Close'])
usd_location = 'USD_ind.csv'
df_usd = pd.read_csv(usd_location,index_col = "Date", parse_dates=True, usecols=['Date', 'Price'])
df = pd.merge(df_xau, df_usd, how='inner',left_index=True, right_index=True)
df.columns = [u'XAUUSD', u'USDIND']
df_daily_return = compute_daily_returns(df)
# print df_daily_return
df_daily_return.plot(kind='scatter', x='XAUUSD', y='USDIND')
beta_USDIND, alpha_USDIND = np.polyfit(df_daily_return['XAUUSD'], df_daily_return['USDIND'], 1)
plt.plot(df_daily_return['XAUUSD'], beta_USDIND*df_daily_return['XAUUSD'] + alpha_USDIND,'-', color='r')
plt.show()
print "beta_USDIND=", beta_USDIND
print "alpha_USDIND=", alpha_USDIND
print "黄金与美元指数相关度:"
print df_daily_return.corr(method='pearson')
print "具体系数："
print "beta斜率=", beta_USDIND
print "alpha因子=", alpha_USDIND
```

相关度在[-1 . 1] 之间， 黄金和美元指数相关度为 -0.4， 说明一个上升，另一个下跌，相关度的绝对值为0.4，为较强相关度。
同时alpha值较小，ps: 最近偶尔听到炒alpha，比如相比起美股，h股，市场上有认为a股中很多活跃股不跟大市，甚至不跟板块。

分析完黄金和美元是相关度较高后，我们开始portfolio的研究，把黄金，美元指数，原油列入我们的portfolio里面，假设总投资100万， 并以黄金，美元指数，原油，分别40%， 40%， 20%在年头第一天进行资产配置。然后看看这笔投资在年底会变成怎样，并分析这样的配置下，各种数据指标， 包括总的年化收益率， 日回报均值，日回报标准差，日回报峰值，以及夏普比率。

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math

init_alloc = np.array([0.4, 0.4, 0.2])
start_val = 100

def compute_daily_returns(df):
	daily_returns = (df / df.shift(1)) - 1
	daily_returns.ix[0, :] = 0
	return daily_returns

def normalize_data(df):
    return df/df.ix[0,:]

def fillter_data_by_date(df, start_date, end_date):
	return df.ix[start_date:end_date]

def calculate_port_val_total_each_day_daily_ret(df_total_one_year):
    df_normalize = normalize_data(df_total_one_year)
    df_allocated = df_normalize * init_alloc
    pos_vals = df_allocated * start_val
    sum_array = np.sum(pos_vals.values, axis=1)
    port_val_each_day_by_date = pd.DataFrame({"daily_ret": sum_array}, index=pos_vals.index.values)
    port_val_total_each_day_daily_ret = compute_daily_returns(port_val_each_day_by_date)
    return port_val_total_each_day_daily_ret

def portfolio_stat(port_val_total_each_day_daily_ret):
    cum_ret = port_val_total_each_day_daily_ret.sum(axis=0)
    print "cum return=", cum_ret[0]
    port_val_total_each_day_daily_ret_mean = port_val_total_each_day_daily_ret.mean()
    print "mean=", port_val_total_each_day_daily_ret_mean[0]
    port_val_total_each_day_daily_ret_std = port_val_total_each_day_daily_ret.std()
    print "std=", port_val_total_each_day_daily_ret_std[0]
    port_val_total_each_day_daily_ret_kurtosis = port_val_total_each_day_daily_ret.kurtosis()
    print "kurtosis=", port_val_total_each_day_daily_ret_kurtosis[0]
    daily_rf = math.pow(1.02, 1.0/252) -1
    sr = (port_val_total_each_day_daily_ret - daily_rf).mean() / port_val_total_each_day_daily_ret_std
    print "Sharpe ratio=", sr[0]
    
xau_location = '10_Y_DXAUUSD.csv'
df_xau = pd.read_csv(xau_location,index_col = "Date", parse_dates=True, usecols=['Date', 'Close'])
usd_location = 'USD_ind.csv'
df_usd = pd.read_csv(usd_location,index_col = "Date", parse_dates=True, usecols=['Date', 'Price'])
df = pd.merge(df_xau, df_usd, how='inner',left_index=True, right_index=True)
df.columns = [u'XAUUSD', u'USDIND']
usd_location = 'WTI_OIL.csv'
df_oil = pd.read_csv(usd_location,index_col = "Date", parse_dates=True, usecols=['Date', 'OilPrice'])
df_total = pd.merge(df, df_oil, how='inner',left_index=True, right_index=True)
df_total.columns = [u'XAUUSD', u'USDIND',u'WTIOIL']
df_total_one_year = fillter_data_by_date(df_total, '20160101', '20170101')

port_val_total_each_day_daily_ret = calculate_port_val_total_each_day_daily_ret(df_total_one_year)
portfolio_stat(port_val_total_each_day_daily_ret)
```
