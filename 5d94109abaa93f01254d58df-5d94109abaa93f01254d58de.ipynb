# 趋势动量因子

在川总的微信公众号看到关于这篇文章的讨论，然后去看了下原文做了个A股版本
A trend factor: Any economic gains from using information over investment horizons (Han et al. 2016)

构造的思路大概是：
1. 在每个月月末，计算股票价格的N期移动平均$A_{i,t}^{N}$，然后除以当前价格$P_{i,t}$。得到不同期的动量$X_{i,t}^{N}$
2. 将下个月的收益率作为Y，进行回归Y~X，得到各个动量的回归系数$\beta_{N,t}$
3. 预测：根据过去12期的系数得到对于下一期各个动量系数的预测值$E(\beta_{N,t})=\sum^{12}\beta_{N,t}/12$，从而得到对于下一期每个股票收益率的预测
4. 按照传统的取TOP和LAST构建多空组合计算因子收益率

```python
import pandas as pd
import numpy as np
from statsmodels.api import OLS,add_constant
import matplotlib.pyplot as plt
```

```python
# 日历。获取日期list和月末日期list
caldate = DataAPI.TradeCalGet(exchangeCD=u"XSHG",beginDate=u"20070721",endDate=u"20190930",
                              field=u"calendarDate,isOpen,isWeekEnd,isMonthEnd,isQuarterEnd",pandas="1")
caldate = caldate[caldate.isOpen == 1].drop('isOpen', axis=1)
date_list = caldate.calendarDate.tolist()
monthEnd = caldate[caldate.isMonthEnd == 1].calendarDate.tolist()
```

原论文用的是移动平均期数更多更长，但是考虑到A股时间没那么长，会有很多股票因此被剔除出样本，所以没用。
另外注意数据的可获性，每次回归需要用到本月收益和上一月末得到的移动平均数据

```python
window_list = [10,20,40,60,120,240]#,400,600,800,1000] # 论文中使用了3,5,10,20,...,1000
Beta = []
avg_data_all = {}
# 直接使用全A股数据
#stks=DataAPI.EquGet(secID=u"",ticker=u"",equTypeCD=u"A",listStatusCD=u"L",field=u"",pandas="1").secID
#stks = stks[stks.str.startswith('60')|stks.str.startswith('3')|stks.str.startswith('0')].tolist()
last_Year = monthEnd[18][:4]
for day in monthEnd[18:]: # 因为后文要取前200个交易日，所以取的交易日历要比开始日期更早一些
    if day[:4] != last_Year:
        print(last_Year+' done.')
        last_Year = day[:4]
    # 使用指数300或500数据
    stks = DataAPI.IdxConsGet(ticker=u"399905",isNew=u"",intoDate=day,field=u"consID",pandas="1").consID #399905
    stks = stks[stks.str.startswith('60')|stks.str.startswith('3')|stks.str.startswith('0')].tolist()
    # 使用流通市值最小的
    last_month = monthEnd[monthEnd.index(day) - 1] # 上一期月末
    pre_100 = date_list[date_list.index(last_month) - 199] # 向前取200天
    #stks = DataAPI.MktEqudGet(secID=u"",ticker=u"",tradeDate=last_month,beginDate=u"",endDate=u"",isOpen="",
    #                          field=u"secID,negMarketValue,chgPct",pandas="1")
    #stks.negMarketValue /= stks.chgPct+1
    #stks = stks.sort_values('negMarketValue',ascending=True).iloc[:800].secID.tolist()
    # 获取价格数据
    data = DataAPI.MktEqudGet(secID=stks,ticker=u"",beginDate=pre_100,endDate=last_month,isOpen="",
                              field=u"secID,tradeDate,closePrice",pandas="1")
    data = data.set_index(['secID','tradeDate']).closePrice.unstack(0)
    avg_price = []
    for w in window_list:
        avg_price.append(data.iloc[-w:].mean(axis=0) / data.iloc[-1])
    avg_price = pd.concat(avg_price, axis=1).dropna()
    avg_price.columns = list(map(lambda x:str(x), window_list))
    avg_data_all[last_month] = avg_price # 这里的price是用day末可以拿到的价格计算的
    # 获取本月的月收益
    ret_nextM = DataAPI.MktEqumGet(secID=stks,monthEndDate=day,field=u"secID,chgPct",pandas="1"
                                  ).set_index('secID').chgPct.loc[avg_price.index]
    # OLS
    res = OLS(ret_nextM, add_constant(avg_price),missing='drop').fit() # 这个是要在next_month才能得到的回归
    Beta.append(res.params.values[1:])
Beta = pd.DataFrame(Beta, index=monthEnd[18:], columns=window_list) # 表示在index时才能获取的回归贝塔
```

```python
# 根据过去12期的beta，预测下一期的beta
Beta_predict = Beta.rolling(12).mean().dropna() # beta预测是当期就可以用的
trend_factor = pd.Series()
for day in Beta_predict.index:
    try:
        avg_price = avg_data_all[day]
    except:
        stks = DataAPI.IdxConsGet(ticker=u"399905",isNew=u"",intoDate=day,field=u"consID",pandas="1").consID.tolist()
        pre_100 = date_list[date_list.index(day) - 200]
        #stks = DataAPI.MktEqudGet(secID=u"",ticker=u"",tradeDate=day,beginDate=u"",endDate=u"",isOpen="",
        #                          field=u"secID,negMarketValue,chgPct",pandas="1")
        #stks.negMarketValue /= stks.chgPct+1
        #stks = stks.sort_values('negMarketValue',ascending=True).iloc[:300].secID.tolist()
        # 获取价格数据
        data = DataAPI.MktEqudGet(secID=stks,ticker=u"",beginDate=pre_100,endDate=day,isOpen="",
                                  field=u"secID,tradeDate,closePrice",pandas="1")
        data = data.set_index(['secID','tradeDate']).closePrice.unstack(0)
        avg_price = []
        for w in window_list:
            avg_price.append(data.iloc[-w:].mean(axis=0) / data.iloc[-1])
        avg_price = pd.concat(avg_price, axis=1).dropna()
        avg_price.columns = list(map(lambda x:str(x), window_list))
    ret_expected = np.dot(Beta_predict.loc[day], avg_price.T)
    r = pd.DataFrame(ret_expected, index=avg_price.index, columns=[day]).stack()
    trend_factor = trend_factor.append(r)
    
# 得到每一期所有股票的预测值。根据这个值做多和做空20%
trend_factor = pd.DataFrame(trend_factor) # 这个是本月可得的avg_price和本月可得的_beta计算出的
trend_factor['secID'] = [x[0] for x in trend_factor.index]
trend_factor['date'] = [x[1] for x in trend_factor.index]
trend_factor.set_index(['secID','date'],inplace=True)

trend_factor_raw = trend_factor.iloc[:,0].unstack(0)
trend_factor = trend_factor_raw.shift(1)
buy_rank = trend_factor.rank(axis=1, ascending=False)
sell_rank = trend_factor.rank(axis=1, ascending=True)
buy_signal = buy_rank.copy()
sell_signal = sell_rank.copy()
buy_signal[:] = np.where(buy_signal < 100, 1, 0)
sell_signal[:] = np.where(sell_signal < 100, 1, 0)
buy_signal = (buy_signal.T / buy_signal.sum(axis=1)).T
sell_signal = (sell_signal.T / sell_signal.sum(axis=1)).T
```

# 计算多空收益率
在计算多空收益率时，我习惯对收益率做一个处理，剔除尾部极端数据。因为1.这种收益很可能来自于特殊事件（长期停牌复牌，爆炒）且难以重演，并且我们也不希望因子的收益率来自于个别股票的特质收益率而非投资组合

## 全A股样本，剔除科创。
取TOP100和LAST100构建的多空

```python
ret_m_all = DataAPI.MktEqumGet(secID=trend_factor.columns,beginDate=trend_factor.index[0],endDate=trend_factor.index[-1],
                               field=u"secID,endDate,return",pandas="1").set_index(['secID','endDate'])['return'].unstack(0)
# 太大的月收益并不是我想要的，这个可遇不可求，难以重复。数值是随便设置的，大概5%的quantile水平
ret_m_all[ret_m_all>0.25] = 0.25
ret_m_all[ret_m_all<-0.2] = -0.2

long_ret = (ret_m_all * buy_signal).sum(axis=1)
short_ret = (ret_m_all * sell_signal).sum(axis=1)

long_nv = (1+long_ret).cumprod()#-1
short_nv = (1+short_ret).cumprod()#-1
relative_nv = (1+long_ret-short_ret).cumprod()

f = plt.figure(figsize=(12,8))
ax = f.add_subplot(111)
#ax1 = ax.twinx()
ax.set_yscale('log')
long_nv.plot(ax=ax,color='r')
short_nv.plot(ax=ax,color='g')
relative_nv.plot(ax=ax,color='b')
ax.legend(['long_nv','short_nv','relative_nv'])
```

# 总结
1. 因子有点失效嫌疑，所以就发出来给大家图一乐，主要在于拓展思路
2. 全文一直在移动平均。无论是计算价格的移动平均，还是预测时对于$\beta$的移动平均
3. 本质是思路还是探究长短期动量的预测能力，和HAR模型有点像？
4. 没有做多组的分位数回测，可以自行尝试，看看单调性如何
5. 动量研究局限二级市场的价格行为，可以考虑搭配选股模型
6. 实际使用里因子还是有很多可以改进的地方，大家可以探究，比如5,6：
7. 短期动量因子（3,5,10日这种的）依旧使用过去12期的平均值作为预测，似乎直觉上说不过去
8. 能不能加入对于不同期因子有效性的检验，排除一些近期连续不显著的因子（如果证明出这种显著性具有持续性），以减小模型的噪声
