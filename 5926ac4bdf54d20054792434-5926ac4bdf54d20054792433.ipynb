# 股票筛选

```python
import pandas as pd

# 默认筛选ROIC近5年大于15%的公司
def getTickersByRatio(list_ticker = [], index = 'ROIC', level = 15, years = 5):
    # 收益率
    df = DataAPI.FdmtIndiRtnGet(ticker=list_ticker,secID="",endDate="",beginDate="",beginYear=u"",endYear=u"",reportType=u"A",field=u"",pandas="1")
    # 筛选years年内收益率指标index大于level的股票
    list_ticker = []
    for ticker in df.groupby('ticker'):
        df_temp = ticker[1][:years]
        if len(df_temp[df_temp[index]>level]) == years:
            list_ticker.append(ticker[0])
    return list_ticker

# 默认筛选近3年净现金流大于0的公司
def getTickersByCash(list_ticker = [], index = 'NChangeInCash', level = 0, years = 3):
    # 现金流
    df = DataAPI.FdmtCFAllLatestGet(ticker=list_ticker,secID=u"",reportType=u"A",endDate=u"",beginDate=u"",year="",field=u"",pandas="1")
    # 筛选years年内收益率指标index大于level的股票
    list_ticker = []
    for ticker in df.groupby('ticker'):
        df_temp = ticker[1][:years]
        if len(df_temp[df_temp[index]>level]) == years:
            list_ticker.append(ticker[0])
    return list_ticker

# 筛选股票，默认根据ROIC筛选
def getStocks(list_ticker, index = 'ROIC'):
    # 初始化输出列表
    df_result = pd.DataFrame(index=list_ticker, columns=['公司名称','近五年平均' + index,'最新统计年度','近五年' + index + '(时间由近到远)','所属行业1','所属行业2','所属行业3'])

    # 收益率
    df_ratio = DataAPI.FdmtIndiRtnGet(ticker=list_ticker,secID="",endDate="",beginDate="",beginYear=u"",endYear=u"",reportType=u"A",field=u"",pandas="1")
    

    # 申万行业分类
    industry = DataAPI.EquIndustryGet(industryVersionCD=u"",industry=u"申万行业分类",secID=u"",ticker=u"",intoDate=u"",field=u"",pandas="1")
    industry = industry.sort(columns=['outDate'], ascending=[False])
    industry = industry.drop_duplicates(['ticker'])

    for ticker in df_ratio.groupby('ticker'):
        df_temp = ticker[1][:5]
        stockcode = ticker[0]
        stockname = df_temp.secShortName.tolist()[0]
        endDate = df_temp.endDate.tolist()[0]
        ratio_avg = df_temp[index].mean()

        ratio_str = ""
        ratio_list = df_temp[index].tolist()
        for ratio in ratio_list:
            ratio_str += str(round(ratio,2))
            if ratio_list.index(ratio) < len(ratio_list) - 1 :
                ratio_str += " | "

        industryName1 = industry[industry['ticker']==stockcode].industryName1.tolist()[0]
        industryName2 = industry[industry['ticker']==stockcode].industryName2.tolist()[0]
        industryName3 = industry[industry['ticker']==stockcode].industryName3.tolist()[0]
        df_result.ix[stockcode] = [stockname, ratio_avg, endDate, ratio_str, industryName1,industryName2,industryName3]

    df_result = df_result.sort(columns=['近五年平均' + index],ascending=[False])
    
    return df_result


# 得到同行业股票，默认根据ROIC由高到低排序，取前十
def getStocksCorr(stockcode , index = 'ROIC'):
    
    # 申万行业分类
    industry = DataAPI.EquIndustryGet(industryVersionCD=u"",industry=u"申万行业分类",secID=u"",ticker=u"",intoDate=u"",field=u"",pandas="1")
    industry = industry.sort(columns=['outDate'], ascending=[False])
    industry = industry.drop_duplicates(['ticker'])
    
    industryID3 = industry[industry['ticker']==stockcode].industryID3.tolist()[0]
    df_industry = industry[industry['industryID3']==industryID3].drop_duplicates(['ticker'])

    list_industryStock = df_industry.ticker.tolist()

    df_industry_ratio = DataAPI.FdmtIndiRtnGet(ticker=list_industryStock,secID="",endDate="",beginDate="",beginYear=u"",endYear=u"",reportType=u"A",field=u"",pandas="1")

    df_sort = pd.DataFrame(index=list_industryStock, columns=[index + '_AVG'])
    for stock in df_industry_ratio.groupby('ticker'):
        stockcode = stock[0]
        df_temp = stock[1][:5]
        ratio_average = df_temp[index].mean()
        df_sort.ix[stockcode] = [ratio_average]
    df_sort = df_sort.dropna()
    df_sort = df_sort.sort(columns=[index + '_AVG'], ascending=[False])
    list_corr  = df_sort.head(10).index.tolist()
    df_corr = DataAPI.EquGet(equTypeCD=u"A",secID=u"",ticker=list_corr,listStatusCD=u"",field=u"",pandas="1")

    
    listTitleE = ['ticker','secShortName','listDate','primeOperating','endDate','TShEquity']
    listTitleC = ['代码','简称','上市日期','主营业务范围','财务报告日期','所有者权益合计']
    df_corr = df_corr[listTitleE]
    for i in range(len(listTitleE)):
        df_corr = df_corr.rename(columns={listTitleE[i]:listTitleC[i]})
    return df_corr

# 主函数
def main(indexRatio = "ROIC", levelRatio = 15, yearsRatio = 5, yearsCash = 3):
    # A股所有股票
    df_A = DataAPI.EquGet(equTypeCD=u"A",secID=u"",ticker='',listStatusCD=u"",field=u"",pandas="1")
    list_ticker = df_A['ticker'].tolist()

    # 筛选指标:收益率 ， 标准（大于）：15% ， 年限：近5年
    list_ticker = getTickersByRatio(list_ticker, index = indexRatio, level = levelRatio, years = yearsRatio)
    # 筛选指标：净现金流 ， 标准（大于）：0 ， 年限：近3年
    list_ticker = getTickersByCash(list_ticker, years = yearsCash)

    # 显示股票和行业信息
    df_stocks = getStocks(list_ticker, index = indexRatio)
    return df_stocks
```

```python
#----------修改该部分内容----------
#可以修改收益率类型、收益率筛选标准、筛选年限
#
# 收益率类型：ROIC、ROA、ROE
indexRatio = "ROIC"
# 收益率筛选标准（不用写百分号"%"）
levelRatio = 15
# 收益率筛选年限
yearsRatio = 5
# 现金流大于0筛选年限
yearsCash = 3
#--------------------------------

main(indexRatio, levelRatio, yearsRatio, yearsCash)
```
