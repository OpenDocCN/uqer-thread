而泽熙官网的数据显示，泽熙一期、二期、三期、四期、五期多只基金的累积净值已经连续2个多月没有大幅变动。

说明泽熙已经空仓一段时间，那么现在泽熙还重仓哪些股票呢？

```python
import pandas as pd
```

```python
a_sec_id = set_universe('A')
a_tic = [x[0:6] for x in a_sec_id]

data = pd.DataFrame()

for i in range(3):
    tmp = DataAPI.EquIndustryGet(ticker=a_tic[i*1000:(i+1)*1000],industryVersionCD=u"010303",field=u"secID,secShortName,partyID,",pandas="1")
    tmp = tmp.groupby('secID').first()
    data = pd.concat([data, tmp])
    
for sec in data.index:
    party_id = data.partyID[sec]
    
    try:
        holder = DataAPI.JY.EquMainshJYGet(partyID=party_id,infoTypeCD=u"2",endDateEnd=u"20150930",field=u"",pandas="1")
    except:
        continue
    holder = holder[holder.endDate == holder.endDate[0]]
    holder['zexi'] = [name.find("泽熙") for name in holder.shName]
    holder = holder[holder.zexi > -1][['endDate','shName','holdVol','holdPct']]
    
    if len(holder.index) > 0:
        print sec
```

可以看到泽熙现在进入了两家公司的前十大流通股东：
