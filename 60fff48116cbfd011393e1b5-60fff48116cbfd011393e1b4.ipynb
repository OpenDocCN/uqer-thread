# Groupby

```python
import pandas as pd
import numpy as np
df = pd.DataFrame({'A': ['a', 'b', 'a', 'c', 'a', 'c', 'b', 'c'], 
                       'B': [2, 7, 1, 3, 3, 2, 4, 8], 
                       'C': [100, 87, 96, 130, 105, 87, 96, 155]})
print(df)
```

```python
print(df.groupby('A').mean())
print(df.groupby(['A','B']).mean())
```

```python
data=df.groupby('A')#分组后，选择列进行计算：
print(data['B'].std())
print(data['B','C'].median())

print(data.agg({'B':'mean','C':'sum'}))#按A进行分组后，可以对不同的列采用不同的聚合方法
dd=df.groupby('A').agg({'B':['mean','sum'],'C':['count','std']})#按照某列进行分组后，对不同的列采用不同的聚合方法。注意：字典的形式就是要用大括号
print(dd)
            
print(df.groupby('A').apply(np.mean))#如果按照A进行分组后，对多列采用相同的聚合方法，我们可以借助apply函数：

ff=df.groupby('A')['B'].agg(['mean','median','sum'])#使用groupby按照某列（A）进行分组后，需要对另外一列采用不同的聚合方法。
print(ff)

            
```

```python
rr=df.groupby('A')['B','C'].transform('max')#应用到所有分组
print(rr)
```
