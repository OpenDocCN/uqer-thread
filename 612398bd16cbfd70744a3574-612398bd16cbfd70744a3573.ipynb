# 可转债双低轮动策略

# 思路
根据集思录上主流的可转债双低策略，每月挑选可转债价格+溢价率最低的20只可转债持有。由于各大量化平台都不支持可转债买卖，因此只有用研究模拟。
# 因子计算
可转债双低得分=可转债价格+溢价率\*100(集思录可查。注意优矿的溢价率是乘以100以后的，即10%溢价率值为10，因此不乘以100)
# 模拟过程
每月计算得分，买入得分最小的20只。下月初计算收益率
# 择时
由于数据量大运行太慢，因此没使用复杂判断。使用市场平均双低得分，高于160时空仓，低于160时满仓。

```python
# 获取某日可转债指数所有成分
def get_bond(tradeDate):
    bond=DataAPI.IdxConsGet(ticker=u"000832",intoDate=tradeDate,field=u"consID",pandas="1")['consID']
    tmp_ID=list(bond.values)
    #去掉可交换债
    cb=DataAPI.BondGet(secID=tmp_ID,typeID=u"02020113",field=u"secID",pandas="1")
    bond=list(cb['secID'])
    #去掉强赎时间少于1年的
    return list(bond)                                                                                 
```

```python
#获取某日的市场平均溢价率
def get_avg(date):
    bond=get_bond(date)#获取所有可转债
    #print date
    bondPremRatio=DataAPI.MktConsBondPerfGet(beginDate=date,endDate=date,secID=bond,
                       field=u"secID,bondPremRatio",pandas="1")
    
    bondPremRatio.set_index('secID',inplace=True)
    bondPremRatio.columns=['value']
    #排除当日停牌的
    price=DataAPI.MktBonddGet(secID=bond,beginDate=date,endDate=date,
                                  field=u"secID,openPrice",pandas="1")
    price=price[price.openPrice>0]
    price.set_index('secID',inplace=True)
    price.columns=['value']
    z=price+bondPremRatio
    z.dropna(inplace=True)
    return z.mean().value
```

# 以下是模拟策略

```python
from lib.functions import *
#获取每月最后一个交易日
cal=get_month_last_tradeDate('20121201','20210801')
```

# 方案1：买入双低（价格+溢价率）最低的20只，每月轮动，无择时。

```python
result={}
to_hold=[]
for i in range(len(cal)):
    if i!=0:
        #获取本月收益
        if len(to_hold)>0:
            price=DataAPI.MktBonddGet(secID=to_hold,beginDate=cal[i-1],endDate=cal[i],
                                      field=u"secID,tradeDate,openPrice,closePrice",pandas="1")
            price=price.groupby(['secID','tradeDate']).mean().unstack().T
            openPrice=price.loc['openPrice'].iloc[0]
            closePrice=price.loc['closePrice'].iloc[-2]#调仓日前一天卖出
            #print openPrice,closePrice
            result[cal[i]]=(closePrice/openPrice-1).mean()
        else:
            result[cal[i]]=0
    bond=get_bond(cal[i])#获取所有可转债
    #获取可转债溢价
    bondPremRatio=DataAPI.MktConsBondPerfGet(beginDate=cal[i],endDate=cal[i],secID=bond,
                       field=u"secID,bondPremRatio",pandas="1")

    bondPremRatio.set_index('secID',inplace=True)
    bondPremRatio.columns=['value']
    #排除当日停牌的
    price=DataAPI.MktBonddGet(secID=bond,beginDate=cal[i],endDate=cal[i],
                                  field=u"secID,openPrice",pandas="1")
    price=price[price.openPrice>0]
    price.set_index('secID',inplace=True)
    price.columns=['value']
    z=price+bondPremRatio
    z.dropna(inplace=True)
    z.sort_values(by='value',ascending=True,inplace=True)
    to_hold=list(z.iloc[:20].index)   
```

```python
import pandas as pd
(pd.Series(result)+1).cumprod().plot()
```

## 结果分析
从2013年到2021年7月30日，策略收益165%，年化收益率12%。但是其中最大回撤也很大，从2015年最高点到2018年的最低点，回撤50%以上，完全没有体现出低风险投资的好处。

# 方案2：买入双低（价格+溢价率）最低的20只，每月轮动，且市场平均双低高于160时不买入。

```python
result={}
to_hold=[]
for i in range(len(cal)):
    if i!=0:
        #获取本月收益
        if len(to_hold)>0:
            price=DataAPI.MktBonddGet(secID=to_hold,beginDate=cal[i-1],endDate=cal[i],
                                      field=u"secID,tradeDate,openPrice,closePrice",pandas="1")
            price=price.groupby(['secID','tradeDate']).mean().unstack().T
            openPrice=price.loc['openPrice'].iloc[0]
            closePrice=price.loc['closePrice'].iloc[-2]#调仓日前一天卖出
            #print openPrice,closePrice
            result[cal[i]]=(closePrice/openPrice-1).mean()
        else:
            result[cal[i]]=0
    if get_avg(cal[i])<160:#平均双低大于160不买入
        bond=get_bond(cal[i])#获取所有可转债
        #获取可转债溢价
        bondPremRatio=DataAPI.MktConsBondPerfGet(beginDate=cal[i],endDate=cal[i],secID=bond,
                           field=u"secID,bondPremRatio",pandas="1")

        bondPremRatio.set_index('secID',inplace=True)
        bondPremRatio.columns=['value']
        #bondPremRatio.sort_values('bondPremRatio',ascending=True,inplace=True)
        #bondPremRatio=zscore(bondPremRatio)
        #排除当日停牌的
        price=DataAPI.MktBonddGet(secID=bond,beginDate=cal[i],endDate=cal[i],
                                      field=u"secID,openPrice",pandas="1")
        price=price[price.openPrice>0]
        price.set_index('secID',inplace=True)
        price.columns=['value']
        #price=zscore(price)
        #print cal[i]
        #print(price,bondPremRatio)
        z=price+bondPremRatio
        z.dropna(inplace=True)
        z.sort_values(by='value',ascending=True,inplace=True)
        to_hold=list(z.iloc[:20].index)
    else:
        to_hold=[]    
```

```python
(pd.Series(result)+1).cumprod().plot()
```

## 测试结果
最终总收益240%，年化收益率15.4%,最大回撤15.7%。说明择时效果不错。

# 注意事项

- 在牛市后期（2015年）本策略过早下车，没有吃到后面的红利。现在也有这种顾虑，近期市场平均双低值一直高于160，只有眼睁睁看着他涨。但是盈亏同源，如果把阈值设为170总收益反而会降低，且最大回撤增加很多。

- 可转债历史净值与正股偶尔不同步，比如2017年反而出现回撤，在2018年回撤较小，可以作为正股对冲。

- 以上策略看上去计算复杂，好在有集思录给我们提供了现成的平均双低数据和各可转债双低值，每天自己查询即可。

- 回测中没有考虑强赎的问题，实际持仓时必须随时注意持有的可转债遇到强赎公告必须马上卖出或转股，以免较大损失。
