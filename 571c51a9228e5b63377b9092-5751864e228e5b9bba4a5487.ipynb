# 简单策略2（时间修正版)【代码小白一枚，但自觉对市场感悟还不错】

通过对600570的简单回测，第一笔交易在2月16日开始。可以看出短期效果还是不错，在基准收益为-18%的同时，策略收益达到71%。缺点：股价上升阶段，吃到的利润还有不足，比如在2015年4——6月牛市顶部，基准最高年化达到144%，而策略最高85.6%，接近60%个点，肉还是疼。优点：股灾下跌期间，粗糙但较好地避开股灾，并在每一次底部抢到反弹。
缺陷2：盘整期间，并不比基准有更大优势，反而频繁交易与错误节奏导致出现交易性损耗。

拉长时间进行测试。

从2008年年初到2016年4月，可以看出还是长时间跑赢指数，年化达到45%。慢牛期间能显著跑赢基准，但大牛行情中，由于过于谨慎，导致获利不多。

再换一只股票进行测试。

跑赢基准，牛市上升行情中，有一段与基准几乎一样，还是蛮奇特。继续拉长时间。

这TM就尴尬了，为什么会有这么长一段时间，跑输基准？

把时间换到前面几年看一看。

发现从2008年到2010年，震荡得更加厉害，导致无法有效避开阴跌，上涨期间无法积累盈利，加之盘整期间出现节奏+交易损耗。

继续改……

删掉最后一条代码，发现无论哪只票，策略效益都减少了。证明这条止损指令存在不可替代的意义。下面试试能不能优化它。

修改最后一条代码，感觉业并没有多大区别……这TM就尴尬了……

仔细想想，最后一条代码，实际上是为倒数第二条上的保险，针对的，当然是下跌时候的情景，ma30其实也是包含在ma60中，所以没有多大改善。但这里面，似乎缺了20——60这个区间段的对应操作。20日线上，10阶滞后之上可以持股，这是牛市。但20——60，这就是标准反弹了。止损不缺，缺止盈了。

这TM又尴尬了……改60后，这刷新高能力又下降了，不够坚挺啊！不过好歹年化还是增长了1个百分点。

喜忧参半……但大致发现，30易创新高，反弹快；60则较稳定，短期冲击力不强。

多加一条止盈策略，很明显，对风险控制不明显，但对降低收益的负面影响却很大。
