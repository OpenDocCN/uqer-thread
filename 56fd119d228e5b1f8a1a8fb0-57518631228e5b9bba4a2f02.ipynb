### 1、优矿上的财报
---
- 优矿上所有的财报数据都在【数据】——【基本面数据】下，包括常见的三大报表以及业绩预告表

- 但使用者会发现，对于某一张表会出现很多细分，以利润表为例，就会发现在API中有如下几个与利润表相关的API
    - 证券业利润表（所有会计期末最新披露）、证券业利润表
    - 保险业利润表（所有会计期末最新披露）、保险业利润表
    - 银行业利润表(所有会计期末最新披露)、银行业利润表
    - 一般工商业利润表(所有会计期末最新披露)、一般工商业利润表
    - 合并利润表、合并利润表(最近)、合并利润表(所有会计期末最新披露)


- 可以很清楚的看到，有两个层次的分类，首先是按照行业的分类，因为金融行业经营的特殊性，比如银行的主要收入来源就是利息差，但是一般工商业则是销售商品或者提供劳务，所以有必要需要将金融业和一般工商业分开来，而且证券、保险、银行都有自己的特有的报表模板，当然所有企业大部分科目都是一致的，所以在合并报表中都能查到所有企业

- 另一个分类层次在于括号后面的东西，这一点就需要对财务数据有一定的了解，以合并报表来说，包括`合并利润表、合并利润表(最近)、合并利润表(所有会计期末最新披露)`，他们分别代表不同的情况，下面用实例来说明（PS：可以在数据栏下搜索“合并利润表”得到这三个API）


### 2、合并利润表
---
- 根据文档描述，合并利润表收集了2007年以来沪深上市公司定期报告中各个会计期间的利润表数据，而且若上市公司对外财务报表进行更正，调整，均有采集并对外展示，`也就是说，这张表是最全面的，将每次更新的数据都记录在册`

- 可能有人对上面标红色的部分还不太理解，简单说明一下，众所周知，每一期的利润表都会公布两个值：当期值+上期值，所以以14年的年度净利润除了在14年的年报上以“当期值”公布外，还会在15年的年报上以“上期值”公布，这样对同一个值就会有两次公布，尽管在绝大多数时候这两个值是相同的（毕竟反映的都是14年的净利润），但也有不同的时候，而且更有甚者，公司今天发了财报，然后过几天又发一个更新版的财报，这样对于同一个数据可能在不同时间有多次公布

- 合并利润表就是记录了所有次的更新，所以说是最全面的

- 以格力电器000651为例，详细说明如下，下例取了格力电器20140101至今的所有财务数据

```python
DataAPI.FdmtISGet(ticker=u"000651",beginDate=u"20140101",field=u"secShortName,endDate,endDateRep,publishDate,reportType,fiscalPeriod,tRevenue",pandas="1")
```

#### 说明几点：
- 以11、12行来看，也就是14年的一季报，可以看到有两条记录，两条记录的不同之处在于`publishDate`不同，分别是14年和15年，也就是说14年的一季度的数据在14年一季报上有公布（作为当期值），同时也在15年的一季报上有公布（作为上期值），这和之前所描述的是一致的

- 细心地读者可能也发现了，两条记录对应的`endDateRep`也不一样，一个是14年一个是15年，结合API中对该字段的描述：报表会计期末，以及上一条的解释，这里不难看出这里指的就是该条数据是在哪一年的报表上出现的

- 另外，还有一个需要注意的是`reportType`中的Q3和CQ3，一个是单季度的（三季度），一个是前三季度的累计值，所以对应的`fiscalPeriod`分别为3和9，这一点在API中的reportType字段也是有介绍的

### 3、合并利润表(所有会计期末最新披露)
---
- 先看API文档描述：收录每一会计期间最新披露的财务数据

- 结合上面的分析，这里应该不难理解了，正是由于一个数据（比如14年一季度的净利润）存在多次披露，而且有时前后披露会有差异，那么为了研究的方便，需要将形成一个一一对应的关系，也就是一个数据只对应一条记录，那么选取哪一条数据呢？根据描述，选取的是最新披露的，也就是publishDate离现在最近的那一条数据，还是以前面的例子来说，14年一季度的数据用的是publishDate为2015-04-30这一条记录，当然这种做法可能会说用到未来数据（在前后数据不一致的时候），所以使用者需要根据自己的实际情况来选取对应的API，以及根据自己的需求来自定义处理方法

- 所以用这个API取数据，得出的结果就是一个数据就是一条记录，一一对应，如下所示

```python
DataAPI.FdmtISAllLatestGet(ticker=u"000651",beginDate=u"20140101",field=u"secShortName,endDate,endDateRep,publishDate,reportType,fiscalPeriod,tRevenue",pandas="1")
```
