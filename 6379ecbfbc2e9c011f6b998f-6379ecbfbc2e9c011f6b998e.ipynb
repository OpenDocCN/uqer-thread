自动分段最小二乘法

先给定一个段长的范围（本实验中段长d为4-8），评价拟合效果的拟合优度[5]为r2，然后将所提供的股票数据依次取段长d个数据，对着几组数据通过一般最小二乘法进行多项式曲线拟合，计算拟合优度 （y为真实值，yc为估计值，为平均值），取max（r2）对应的一组数据为此次拟合的样本值，这次得到的模型就是第一段曲线的模型。将从这段曲线的最后一点开始重复进行前面所需的拟合和判断，直到所有点拟合完毕，得到所有分段曲线的模型。

从最新股价到前面的序列的线性趋势，选R2(rsquared)比较大对应的序列比较现实，但道理不充分，选[r,p]=stats.pearsonr(x,y),直到p<0.05改变不行

```python
df=DataAPI.MktEqudAdjGet(ticker=u"300158",beginDate="20221001",endDate=u"20221118",isOpen="1",
                      field=u"",pandas="1")
df=df[["tradeDate","ticker","secShortName","closePrice"]]
#df=df.sort_values(by=["tradeDate"], ascending=False)
df=df.reset_index(drop=True)
#df=df.tail(10)
df=df.reset_index(drop=True)
df["no"]=df.index
print df
```

```python
import statsmodels.api as sm
import scipy.stats as stats
import pandas as pd
from matplotlib import pyplot as plt
#从最近往前推 线性趋势的断点位置
def get_breakpoint(df,length=30):
    res=pd.DataFrame(columns=["n","r","p","fit_rsquared","no"])
    i=0
    for n in range(5,length):
        df0=df.tail(n)
        #df0=df0.reset_index(drop=True)
        #df0["no"]=df0.index
        x=df0["no"].values
        no=x[0]
        y=df0["closePrice"].values
        r,p=stats.pearsonr(x,y)
        fit=sm.formula.ols('closePrice~no',data=df0).fit()
        #选rsquared比较大 对应的n
        res.loc[i]=[n,r,p,fit.rsquared,no]  
        i=i+1
    return int(res.loc[res["fit_rsquared"].idxmax(),"no"])
#print df
breakpoint=[]
df1=df
m=len(df)
#print m
while m>5:
    m=get_breakpoint(df1,length=30)
    #print m
    breakpoint.append(m)
    df1=df[0:m+1]
#后面加0
breakpoint.append(0)
#倒排
breakpoint= breakpoint[::-1]
#后面加df最后一行的列号
breakpoint.append(len(df)-1)
print breakpoint
```

```python
from matplotlib import pyplot as plt
#plt.scatter(x,y,alpha = 0.3)
fig = plt.figure(figsize=(10,4))
fig.set_tight_layout(True)
fig.set_size_inches((20, 16))
#   [left, bottom, width, height]
#   left=距离图形左侧多远（如x值）
#   bottom=距离图形底部的距离（如y值）
#left, bottom, width, height = 0.1, 0.1, 0.8, 0.8
#figure的百分比,从figure 10%的位置开始绘制, 宽高是figure的80%
ax = fig.add_axes((0, 0.1,0.8, 0.4),axis_bgcolor='white')   
ax.grid(True)
ax.set_title(u'closePrice',fontsize=15, loc='left', color='r')
ax.set_xticks(range(0,len(df), 5))#X轴刻度设定 每5天标一个日期
# X-轴每个ticker标签都向右倾斜45度
for label in ax.xaxis.get_ticklabels():
        label.set_rotation(45)
        label.set_fontsize(10)  # 设置标签字体
ax.set_xticklabels([df.ix[index, 'tradeDate'] for index in ax.get_xticks()])
j=0
for i in breakpoint:
    #ax.plt.xlabel('no')
    #ax.plt.ylabel('closePrice')
    #plt.xticks(df["tradeDate"],rotation=90)#设置时间标签显示格式
    if j!=0:
        start=breakpoint[j-1]
        end=breakpoint[j]
        df0=df[start:end+1]
        y = list(df0.closePrice.values)
        x =list(df0.no.values)
        X = sm.add_constant(x)#给自变量中加入常数项
        model = sm.OLS(y,X).fit()
        ax.plot(df0["no"].values,df0["closePrice"].values,"o",color="blue",linestyle='-.')
        y_hat = model.predict() #获得拟合值
        ax.plot(x,y_hat,'red',alpha =0.9)
        
    j=j+1
    #隐藏坐标轴
ax.get_xaxis().set_visible(True)
ax.get_yaxis().set_visible(True)
fig.show()
```
