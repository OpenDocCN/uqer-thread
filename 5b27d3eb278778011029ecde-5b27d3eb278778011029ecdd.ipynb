## 课程作业:
1. 补充流通市值加权的组合构建方式？
2. 组合优化当中，股票收益率的协方差矩阵估计的方式有哪些，哪种更好？

```python
1. 补充流通市值加权的组合构建方式？
等权加权，所有因子的权重相同
IC均值加权，利用过去一段时间的IC均值为权重进行加权
IC_IR加权，利用过去一段时间的IC均值除以标准差为权重进行加权
最优化复合IC_IR加权，利用过去一段时间的I最优化复合的IC_IR为权重进行加权
半衰期IC加权，对IC序列进行半衰期加权后最为权重进行加权
2. 组合优化当中，股票收益率的协方差矩阵估计的方式有哪些，哪种更好？
压缩估计
随机矩阵模型
因子模型
基于历史数据的模型
随机矩阵模型好一些，根据随机矩阵理论，当股票数量相对于样本数量较大时，位于一定范围内的协方差矩阵特征根与完全随机的收益序列的协方差矩阵特征根相近。因此，通过对该特定范围内的特征根进行修正，可以提高样本协方差矩阵所包含的信息量。
```
