把通联的数据库看成一个黑匣子，数据接口便是一个巨大的遥控器，每一个接口都是遥控器上的一个按键，负责某一个特定的功能。而且这个遥控器还可以不断扩大。
我们以平台中的“沪深股票日行情”接口为例DataAPI.MktEqudGet,提供请求字段和可返回字段。

请求字段：你需要输入函数的自变量，告诉函数特定的需求
可返回字段：你需要数据库给你返回的因变量

这里我们希望得到工商银行（601398）的日度收盘价数据，简单而言如下：（field便是你希望返回的因变量，这里首先返回证券代码，交易日，收盘价三个字段）

```python
data = DataAPI.MktEqudGet(ticker = u'601398',field = ['secID','tradeDate','closePrice'],pandas='1')
print data
```

数据接口返回了工商银行自上市以来的所有日度收盘价。
这里我们甚至发现了一个小技巧：获得某只股票上市首日（当然有别的借口可以直接获得啦）
直接调用tradeDate字段的第一行即可

```python
print data['tradeDate'][0]
```

如果希望得到某段特定时间的行情数据呢？
比如：2015年初至今，工商银行的开／高／低／收

```python
from CAL.PyCAL import *
start_date = Date(2015,1,1)
end_date = Date.todaysDate()
#数据接口中的输入量beginDate，endDate需要字符串形式，所以先转换
start_date = start_date.toDateTime().strftime('%Y%m%d')
end_date = end_date.toDateTime().strftime('%Y%m%d')
#然后调用数据接口
data2 = DataAPI.MktEqudGet(ticker = u'601398',beginDate = start_date, endDate = end_date, field = ['secID','tradeDate','openPrice','highestPrice','lowestPrice','closePrice'],pandas='1')
print data2
```

以上获得的事某一只股票的时间序列数据。
但也许你只想得到某一天市场中所有股票的行情面板数据
那么只需要在输入中做一下修改即可，具体数据参数请参考”数据接口文档“

```python
certain_date = DateTime(2015,11,23)
certain_date = certain_date.strftime('%Y%m%d')  #其实不需要toDateTime()也可以

data3 = DataAPI.MktEqudGet(tradeDate = certain_date, field = ['secID','secShortName','tradeDate','closePrice'])

print data3
```
