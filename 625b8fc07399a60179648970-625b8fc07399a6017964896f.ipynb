# 简单的方法函数_Vide

```python

import numpy as np
from scipy.stats import norm

# 辅助用子函数 更正round()遇0.5时奇数会舍弃0.5的错误



def My_round(x, y=2):
    x = float(x)
    # 指数运算 不是 ^ 而是 **
    if round(x - int(x * 10 ** y) / 10 ** y, y * 3 + 1) == 0.5 / 10 ** y:
        return (round(x + 0.1 / 10 ** (2 * y), y))
    else:
        return (round(x, y))


# 类对象声明
class Atest( object ):
    # 类实例化的时候 的初始值 在 __init__ 里面要求
    def __init__( self ,r,year):
        #
        self.__r = r
        self.__year = year

    # =============================================================================
    # BS()为BS公式的欧式期权定价
    # 参数含义:
    # S=2.382            #标的股票价格
    # K=2.4              #期权执行价
    # Vol=0.2225         #波动率
    # call_put='call'    #期权的购沽类型 '认购'
    # q=0                #标的分红率
    # T=31               #期权剩余的行权期
    # =============================================================================
    def BS( self, S, K, Vol, call_put, T, q=0):
        tau = float(T) / self.__year
        d1 = (np.log(S / K) + (self.__r - q + Vol ** 2 / 2) * tau) / (Vol * np.sqrt(tau))
        d2 = d1 - Vol * np.sqrt(tau)
        if call_put == 'call' or call_put == '认购':
            CP_BS = S * np.exp(-q * tau) * norm.cdf(d1) - K * np.exp(-self.__r * tau) * norm.cdf(d2)
        else:
            CP_BS = K * np.exp(-self.__r * tau) * norm.cdf(-d2) - S * np.exp(-q * tau) * norm.cdf(-d1)
        return (CP_BS)



# py文件 单独调用时 会判断如下 进行运行
# 作用是 当在其他文件import 该py时 if后的条件不会运行
# 但单独调用时会运行
if __name__ == '__main__':

    print(round(3.55,1))
    print(round(3.65,1))
    print(My_round(3.55, 1))
    print(My_round(3.65, 1))

    # 类调用
    Aob =  Atest(r=0.03,year=365)
    # 函数(方法)调用
    V1 = Aob.BS(S=2.852, K=2.65, Vol=0.2452, call_put='put', T=27)

    # 输出需要print()
    print(V1)

    #函数嵌套
    print(My_round(Aob.BS(S=2.852, K=2.65, Vol=0.2452, call_put='put', T=27),4))
```
