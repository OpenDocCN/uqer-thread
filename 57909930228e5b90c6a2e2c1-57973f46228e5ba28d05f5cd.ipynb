## <font color=gray>模型简介</font>
 * #### FED模型是一种在投资界广为使用的资产估值模型。基础模型将股票市场的收入价格比$E/P$（PE的倒数）与长期（如10年期）国债收益率对比，并认为前向一年期收入价格比应当等于长期国债收益率。即 $E/P = Y.$      

## <font color=gray>起源</font>  
 * #### FED模型于90年代在华尔街得名，值得注意的是，虽然被称为FED，这一模型却没有得到美联储董事会的官方认证。早在1997年7月，美联储的"Humphrey-Hawkins Report"给出了1982至1997年间长期债券回报与S&P股票前向收益回报（$E/P$）之间的密切关系，见下图：
 ![](http://i.investopedia.com/inv/articles/site/fedmodel1%20copy.gif)
 * #### 此后不久，在1997至1999年之间，Deutsche Morgan Grenfell的Ed Yardini公布了更多分析股债回报关系的报告，他将这种关系称为“FED股票估值模型”。

## <font color=gray> 支持依据</font>
* ### `资产竞争：`股票和债券两种资产在某种情况下是互为替代品的，因此当股票收益高于债券收益时，人们将更多地投资股票，资金从债市向股市转移，直到模型预测的平衡状态被达到；
* ### `现金流估值：` 现金流估值法认为股票的价格应该等于未来现金流的折现价，国债利率可以被视为无风险利率的替代值，当这一利率下降时，股价$P$上升，因此$E/P$下降；
* ### `Just Look at the Data !` 1965年 - 2001年S&P 500 $E/P$与美国10年期国库券利率关系如图（来自[Asness, 2003](http://down.cenet.org.cn/upfile/36/2005714112330119.pdf) ）：  
![](https://c.hime.io/images/wZAXroE4.png)
  
## <font color=gray> 不足之处</font>  
 * #### 缺乏理论支持，在分析框架中没有公司债的位置；
 * #### 可能存在设定偏误，考虑单个公司时，应该有${\frac  {E_{{{\text{x}}}}}{P_{{{\text{x}}}}}}=R_{{{\text{x}}}}\ [1-T]$，$E_x$表示$x$公司的收入，$P_x$表示$x$公司股票市值，$R_x$表示$x$公司公司债利率，$T$表示公司税率。

* 详情请见[wiki - FED Model](https://en.wikipedia.org/wiki/FED_model)、[investopia - Breaking Down FED Model](http://www.investopedia.com/articles/stocks/08/fed-model.asp)

## <font color=gray> 历史预测效果</font>  
* #### [Maio, 2013](https://ideas.repec.org/a/oup/revfin/v17y2013i4p1489-1533.html) 采用股票收益价格比与长期债券收益之差(* yield gap *)作为市场预测因子，与一些传统预测因子对比，如期限结构溢价、违约风险溢价、利率相对水平。发现* yield gap *有更为显著的预测能力，并且在回归中同时加入以上因子时，也只有* yield gap *因子系数显著。
* #### [Asness, 2003](http://down.cenet.org.cn/upfile/36/2005714112330119.pdf) 则认为FED模型没有多少解释力，他认为PE在市场中往往被高估，而Y（债券收益）是名义变量，将两者放在一起比较没有意义。1965年之后两者相关度高只是因为市场上使用这种模型的投资者多，人为地使PE偏离正常水平。如果从1926年开始看S&P 500 $E/P$与美国10年期国库券利率关系将得到：

```python
# built-in
from __future__ import division
import datetime
import time 
import pandas as pd
import numpy as np
import math
import string

# 3rd-party
import matplotlib
import matplotlib.pyplot as plt
from matplotlib import dates
matplotlib.style.use('ggplot')
import seaborn as sns
sns.set_style('white')
from matplotlib import rc
rc('mathtext', default='regular')
from CAL.PyCAL import *
from sklearn import linear_model

# personal lib
# from lib.time_processing import *
# 直接把lib里的函数放在这啦，方便大家运行
def daysInAMonth( year = 2016,  month = 1 ):
    leapYear = (year % 4 == 0)
    Feb = 28 + leapYear
    monthList = [31, Feb, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
    return monthList[month -1]

def firstDayOfLastMonth(year = 2016, month = 1):
    lastMonth = 12 if month == 1 else month - 1
    rtnYear = year - 1 if month == 1 else year
    return datetime.datetime(rtnYear, lastMonth, 1).strftime('%Y-%m-%d')

def lastDayOfLastMonth( year = 2016, month = 1):
    daysInLastMonth = daysInAMonth(year, 12 if month == 1 else month - 1)
    rtn = datetime.datetime.strptime(firstDayOfLastMonth(year, month), '%Y-%m-%d') + datetime.timedelta(daysInLastMonth - 1)
    return rtn.strftime('%Y-%m-%d')

# warm up
DataAPI.settings.cache_enabled = True
fsize = (12, 12*0.618)
stock_index = '000001'
```

```python
# 相关数据获取及处理函数
def dateIdxEP(idx = '000016', tradeDate = '20160720'):
    # 计算某指数交易日EP值（注意Earning是年化的，因此这里获取的EP是年化的）
    td = datetime.datetime.strptime(tradeDate, '%Y%m%d')
    idxCons = DataAPI.IdxCloseWeightGet(ticker = idx, beginDate = firstDayOfLastMonth(td.year, td.month), 
                                                          endDate = lastDayOfLastMonth(td.year, td.month), field = 'consID,weight')

    secPE = DataAPI.MktStockFactorsOneDayGet(tradeDate, secID = idxCons.consID.values, field = 'secID,PE,LCAP')
    secPE['CAP'] = np.exp(secPE['LCAP'])
    secPE['EP'] = 1 / secPE['PE']
    
    forCpt = idxCons.merge(secPE, left_on = 'consID', right_on = 'secID')
    forCpt.set_index('secID', inplace = True)
    
    colNumerator = ['EP', 'weight', 'CAP']
    colDenominator = ['weight', 'CAP']
    
    indexEP = forCpt[colNumerator].prod(1).sum() / forCpt[colDenominator].prod(1).sum() if forCpt[colDenominator].prod(1).sum() > 0 else np.NAN 
    # (index total earnings)/(index total market value)
    return indexEP


def rangeIdxEP(idx = '000016', beginDate = '20160701', endDate = (datetime.datetime.today() - datetime.timedelta(1)).strftime('%Y%m%d')):
    # 计算某段时间内某指数EP值，仍年化
    x = {}
    cal = DataAPI.TradeCalGet('XSHG', beginDate, endDate, 'calendarDate,isOpen', '1').set_index('calendarDate')
    for date in cal.index:
        if cal[date:date].isOpen[0] == 1:
            tmpDate = datetime.datetime.strptime(date, '%Y-%m-%d').strftime('%Y%m%d')
            x[date] = dateIdxEP(idx, tmpDate)
    return pd.DataFrame(x, index = ['EP_' + idx]).T
```

```python
# 获取数据
raw_data1 = rangeIdxEP(idx = '000001', beginDate = '20070101')     # 沪指 E/P
raw_data2 = rangeIdxEP(idx = '000010', beginDate = '20070101')   # 上证180 E/P
raw_data3 = rangeIdxEP(idx = '000016', beginDate = '20070101')   # 上证50 E/P
raw_data4 = DataAPI.ChinaMacroDataGet(indicID= 'M120000051', beginDate = '20070701', field = 'periodDate,dataValue'
                         ).rename_axis(dict(periodDate = 'tradeDate', dataValue = 'YTM'), axis = 1
                         ).set_index('tradeDate') /100                                         # 10年国债YTM


return1 = DataAPI.MktIdxdGet(ticker = '000001', beginDate = '20070101', field = 'tradeDate,CHGPct').set_index('tradeDate').rename(columns = dict(CHGPct = 'SH_return')) * 252 
return2 = DataAPI.MktIdxdGet(ticker = '000010', beginDate = '20070101', field = 'tradeDate,CHGPct').set_index('tradeDate').rename(columns = dict(CHGPct = 'SH180_return')) * 252 
return3 = DataAPI.MktIdxdGet(ticker = '000016', beginDate = '20070101', field = 'tradeDate,CHGPct').set_index('tradeDate').rename(columns = dict(CHGPct = 'SH50_return')) * 252 


data = raw_data1.merge(raw_data2, left_index = True, right_index = True).merge(raw_data3, left_index = True, right_index = True).merge(raw_data4, left_index = True, right_index = True).merge(return1, left_index = True, right_index = True).merge(return2, left_index = True, right_index = True).merge(return3, left_index = True, right_index = True)
```

```python
with sns.axes_style("white"):
    sns.jointplot(x = pltdata.YTM, y = pltdata.EP_000001, kind = "reg", color = "blue");
```

* ### `回归分析`  
- #### 模型设定：$R_t=\alpha + \beta({\frac EP}_t - Y_t) + \mu_t$
- #### 预计$\beta$为负值，这代表着当E/P偏离YMT较大时，股票市场被高估，将下跌以回到正常水平
- #### 样本期间：2011-06-15 ~ 2016-07-19
注记：也可以把解释变量滞后一期

* #### 可见回归系数为负，这与预期情况一致，但统计上并不显著。这意味着yield_gap在国内市场上并不是成为一个好的投资信号。

```python
regdata = pd.DataFrame()
regdata['yield_gap'] = pltdata.EP_000001 - pltdata.YTM
regdata['stock_return'] = pltdata.SH_return
with sns.axes_style("white"):
    sns.jointplot(x = regdata.yield_gap, y = regdata.stock_return, kind = "reg", color = "blue")
    
    
# Create linear regression object
regr = linear_model.LinearRegression()
# regr.fit(regdata.yield_gap, regdata.stock_return)
x = [[s] for s in regdata.yield_gap]
regr.fit(x[:-1], regdata.stock_return[:-1])     # 运行到前两天
print 'y = ', regr.intercept_, regr.coef_[0], '* x'
```

## <font color=gray>总结</font>
 * #### FED模型以其简明易用、符合市场的特点在美国资本市场广为流传，成为许多对冲、养老基金的定价基础。本文利用uqer的数据分析了中国这一模型在中国市场的表现，结果却不尽如人意。在中国市场上，指数E/P值远远超过10年期国债YTM，这种情况类似于美国市场20世纪30年代到60年代的表现. 这可能是因为股市被高估，或者投资者要求较高的风险回报；也可能是因为债券市场流动性较低，且国债常常被作为调整货币政策的工具，这些特点使国债价格偏离了正常水平。
 * #### 总之，如果要在国内使用FED模型作为投资的参照，可能要三思而后行了~
