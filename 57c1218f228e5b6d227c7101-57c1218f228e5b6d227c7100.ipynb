```python
# 下面两部分只是为了证明数据有缺失，与本文无关。再13年7月10日的时候除权，12日的时候有现金分红，但是再下面的函数FundDivGet虽然说是“收录了2005年以来的历史数据”，但没有体现出来
df = DataAPI.FundNavGet(dataDate=u"",secID=u"270048.OFCN",ticker=u"",beginDate=u"",endDate=u"",field=u"",pandas="1")
df['endDate'] = pd.to_datetime(df['endDate'])
df[df['endDate'] > '20130701']
```

```python
DataAPI.FundDivGet(secID=u"270048.OFCN",ticker=u"",adjustedType=u"",beginDate=u"",endDate=u"",field=u"",pandas="1")
```

```python
#NAV和ACCUM_NAV的关系,ACCUM_NAV = 历史累计分红 + NAV。注：历史累计分红是当成没有收益的。但事实上，如果分红再投资的话是应该有收益的，所以这个标准不是很客观。
res = df[['NAV','ACCUM_NAV','endDate']].set_index('endDate')
res = (res - res.shift(1)).dropna()
res[(res['NAV'] - res['ACCUM_NAV']).round(5) !=0] # 加round主要是python的计算会出现0.00000000000000000000001这样的情形
```

调整后净值就是利用了分红的另一个选项，分红再投资。说明下，分红是按照除权日的净值，而不是现金分红日的净值再投入的。可以去看分红公告。

然后对于勤快分红的债基，历史越长，复权净值和累积净值的差距就越大。
