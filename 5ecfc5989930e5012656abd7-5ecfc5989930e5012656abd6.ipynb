```python
import math
def getindex(ticker):
    tmp=DataAPI.FundGet(ticker=ticker,etfLof=u"etf",pandas="1")
    index=tmp.loc[0]['idxTicker']
    #print("index:",index)
    return index

def PEG(ticker,bd,ed):
    #test=DataAPI.MktIdxdEvalGet(ticker=ticker,beginDate=bd,endDate=ed,PEType=u"1",pandas="1")
    pegtable=DataAPI.MktIdxdEvalGet(ticker=ticker,beginDate=bd,
                                endDate=ed,PEType=u"1",pandas="1")
    pegtable.loc[0]["secID"]="fuck"
    pegl=[0]*pegtable.shape[0]
    for index in range(pegtable.shape[0]-250):
        peg=pegtable.loc[index]["PEValue"]/((float(pegtable.loc[index+250]["indexIncome"])/pegtable.loc[index]["indexIncome"]-1)*100)
        if peg > 5 :
            peg = 5
        if peg < 0 :
            peg = 5
        pegl[index]=peg
    pegtable["PEG"]=pegl
    #print(pegl)
    #print(pegtable)
    return pegtable
def PEG2(ticker,bd,ed):
    pegtable=DataAPI.MktIdxdEvalGet(ticker=ticker,beginDate=bd,
                                endDate=ed,PEType=u"1",pandas="1")
    pegtable.loc[0]["secID"]="fuck"
    pegl=[0]*pegtable.shape[0]
    for index in range(pegtable.shape[0]-500):
        peg=pegtable.loc[index]["PEValue"]/((math.sqrt(float(pegtable.loc[index+500]["indexIncome"])/
                                              pegtable.loc[index]["indexIncome"])-1)*100)
        if peg > 5 :
            peg = 5
        if peg < 0 :
            peg = 5
        pegl[index]=peg
    pegtable["PEG"]=pegl
    #print(pegl)
    #print(pegtable)
    return pegtable

def PEG3(ticker,bd,ed):
    pegtable=DataAPI.MktIdxdEvalGet(ticker=ticker,beginDate=bd,
                                endDate=ed,PEType=u"1",pandas="1")
    pegtable.loc[0]["secID"]="fuck"
    pegl=[0]*pegtable.shape[0]
    for index in range(pegtable.shape[0]-750):
        peg=pegtable.loc[index]["PEValue"]/((math.pow(float(pegtable.loc[index+750]["indexIncome"])/
                                              pegtable.loc[index]["indexIncome"],1.0/3)-1)*100)
        if peg > 5 :
            peg = 5
        if peg < 0 :
            peg = 5
        pegl[index]=peg
    pegtable["PEG"]=pegl
    #print(pegl)
    #print(pegtable)
    return pegtable

def drawPE(pp,pegx,tricker):
    ppmean=pp["PEG"].mean()
    ppstd=pp["PEG"].std()
    pp["mean"]=ppmean
    pp["ls"]=ppmean-ppstd
    pp["hs"]=ppmean+ppstd
    pp.plot("tradeDate",["PEG","mean","ls","hs"],title=pegx+tricker)
    print(pegx,tricker,ppmean,ppstd)
    
indexlist=["159928","159929","159931","159930"]
for index in indexlist:
    print(index)
    ticker=getindex(index)
    print(ticker)
    pegtable=PEG(ticker,"20130101","20200501")
    pp=pegtable[0:pegtable.shape[0]-250]
    print("PEG1")
    drawPE(pp,"PEG1",ticker)
    pegtable=PEG2(ticker,"20130101","20200501")
    pp=pegtable[0:pegtable.shape[0]-500]
    print("PEG2")
    drawPE(pp,"PEG2",ticker)
    pegtable=PEG3(ticker,"20130101","20200501")
    pp=pegtable[0:pegtable.shape[0]-750]
    print("PEG3")
    drawPE(pp,"PEG3",ticker)

'''
pegtable=PEG2(ticker,"20130101","20200501")
pp=pegtable[0:pegtable.shape[0]-500]
print(pp)
ppmean=pp["PEG"].mean()
ppstd=pp["PEG"].std()
pp["mean"]=ppmean
pp["ls"]=ppmean-ppstd
pp["hs"]=ppmean+ppstd
pp.plot("tradeDate",["PEG","mean","ls","hs"])

print(ppmean,ppstd)
#print(test[(test.tradeDate>"2015-02-02") & (test.tradeDate<"2015-03-02")])
'''
```
