```python
import numpy as np
import pandas as pd
from CAL.PyCAL import *
import matplotlib.pylab as plt
fundTicker = u"163412" #基金代码
dateList = ["2017-03-31","2017-06-30","2017-09-30","2017-12-31"] #持仓公布日，Q1:"yyyy-03-31";Q2:"yyyy-06-30";Q3:"yyyy-09-30";Q4:"yyyy-12-31";
beginDate = "20160101" #历史行情起始日
endDate = "20171231" #历史行情结束日
benchmarkTicker = "000300" #比较基准，沪深300
```

1、基金代码处自定义为需要的公募基金代码；
2、dateList为需要查看的历史持仓公布日，按自己需求修改，在这里我们只看2017年的4期；
3、beginDate和endDate为历史行情对比的起止日；
4、benchamarkTicker为选取的基准的代码，正如鲁迅所说，人的幸福感不在于你自己赚多少钱，而在于你比身边的人多赚多少钱（鲁迅：怎么什么都是我说的）。此处基准暂定为沪深300，2017年炒股亏了钱的朋友不妨把比较基准设置为创业板指试一下（不用谢，为他人着想是手哥的一贯作风，手哥一定会考虑你们的内心承受能力）

```python
#取得基金历史净值和沪深300同期价格
fundNav = DataAPI.MktFunddAdjGet(secID=u"",ticker=fundTicker,beginDate=beginDate,endDate=endDate,field=[u"tradeDate",u"closePrice"],pandas="1")
fundNav = fundNav.set_index(fundNav["tradeDate"])
indexPrice = DataAPI.MktIdxdGet(tradeDate=u"",indexID=u"",ticker=benchmarkTicker,beginDate=beginDate,endDate=endDate,exchangeCD=u"XSHE,XSHG",field=[u"tradeDate",u"closeIndex"],pandas="1")
indexPrice = indexPrice.set_index(indexPrice["tradeDate"])
#plot
ax = (fundNav["closePrice"]/fundNav["closePrice"].ix[0]).plot(figsize=(16,6), label='fundNav')
ax = (indexPrice["closeIndex"]/indexPrice["closeIndex"].ix[0]).plot(figsize=(16,6), color='grey', style='--',ax=ax, label='indexPrice')
ax.legend(loc='upper left')
```

通过对比我们可以看出，该基金在2016年初至2017年底期间，与沪深300的走势相似度是很高的，在这种情况下，手哥也不禁怀疑，这个骗子，不会是随便买了点沪深300的成份股吧，本着一定要想办法让三姨接不住招的想法，手哥决定再从持仓上一探究竟。

```python
holdingDF = pd.DataFrame(index = dateList, columns = ['numofStocks'])
for i in range(len(dateList)):
    Date = dateList[i]
    #取持仓公布日前的最后一个交易日
    dateDF = DataAPI.TradeCalGet(exchangeCD=u"XSHG,XSHE",beginDate=Date.replace("-",""),endDate=Date.replace("-",""),field=[u"prevTradeDate",u"isOpen"],pandas="1")
    tradeDate = dateDF.iloc[0,0].replace("-","") if dateDF.iloc[0,1] == 0 else Date.replace("-","")
    #取得基金持仓
    FundHoldings = DataAPI.FundHoldingsGet(reportDate=u"",secID=u"",ticker=fundTicker,beginDate=u"",endDate=u"",secType="",field=u"",pandas="1")
    FundHoldingsQ = FundHoldings[FundHoldings['reportDate']==Date].reset_index(drop=True)
    #c持仓股票数量和总权重
    holdingDF.ix[i,'numofStocks'] = FundHoldingsQ.shape[0]
    holdingDF.ix[i,'weightofStocks'] = np.sum(FundHoldingsQ['ratioInNa'])
#plot
fig = plt.figure(figsize=(16,6))
font.set_size(10)

ax = fig.add_subplot(121)
ax = holdingDF["numofStocks"].T.plot.bar(ax=ax)
ax.set_xticklabels(holdingDF.index,rotation=0,fontproperties=font);
ax.set_ylabel(u'公布持仓股票数量（只）',fontproperties=font);
ax = fig.add_subplot(122)
ax = holdingDF["weightofStocks"].T.plot.bar(ax=ax)
ax.set_xticklabels(holdingDF.index,rotation=0,fontproperties=font);
ax.set_ylabel(u'公布持仓股票权重（%）',fontproperties=font);
```

从公布的持仓数据上来推测，该基金在股票上的权重应该是很高的，半年期披露的详细数据显示股票仓位高达90%，其他时期应该也会大幅高于披露的部分数据，持仓数量如果稳定的话可能保持在60只左右，数量压缩到这么少的话，看来情况并不像简单买点成份股那么肤浅呢。
那就先从公布数据的行业上来分析一下吧（别急，让招式先飞一会儿）

```python
#获取所有申万一级行业
industryDF = DataAPI.EquIndustryGet(industryVersionCD=u"010303",industry=u"",secID=u"",ticker=u"",intoDate=beginDate,field=u"industryName1",pandas="1").drop_duplicates().reset_index(drop=True)
FundIndustrySum = pd.DataFrame(index = dateList,columns = industryDF['industryName1'])
BenchmarkIndustrySum = pd.DataFrame(index = dateList,columns = industryDF['industryName1'])
for i in range(len(dateList)):
    Date = dateList[i]
    #取持仓公布日前的最后一个交易日
    dateDF = DataAPI.TradeCalGet(exchangeCD=u"XSHG,XSHE",beginDate=Date.replace("-",""),endDate=Date.replace("-",""),field=[u"prevTradeDate",u"isOpen"],pandas="1")
    tradeDate = dateDF.iloc[0,0].replace("-","") if dateDF.iloc[0,1] == 0 else Date.replace("-","")
    #取得基金持仓
    FundHoldings = DataAPI.FundHoldingsGet(reportDate=u"",secID=u"",ticker=fundTicker,beginDate=u"",endDate=u"",secType="",field=u"",pandas="1")
    FundHoldingsQ = FundHoldings[FundHoldings['reportDate']==Date].sort_values(by="holdingTicker",axis = 0,ascending = True).reset_index(drop=True)
    #取得持仓股票的行业数据
    FundIndustry = DataAPI.EquIndustryGet(industryVersionCD=u"010303",industry=u"",secID=u"",ticker=FundHoldingsQ['holdingTicker'],intoDate=tradeDate,field=[u"ticker",u"industryName1"],pandas="1")
    FundIndustry = FundIndustry.where(FundIndustry.notnull(), 0)
    FundIndustry['weight'] = FundHoldingsQ['ratioInNa']
    #计算基金的行业分配权重
    for industry in industryDF["industryName1"]:
        FundIndustrySum.ix[i,industry] = np.sum(FundIndustry[FundIndustry["industryName1"]==industry]["weight"])/np.sum(FundIndustry["weight"])
    #取得基准指数的成份股
    benchmark = DataAPI.mIdxCloseWeightGet(secID=u"",ticker=benchmarkTicker,beginDate=tradeDate,endDate=tradeDate,field=u"",pandas="1").sort_values(by="consTickerSymbol",axis = 0,ascending = True).reset_index(drop=True)
    #取得指数成份股的行业数据
    BenchmarkIndustry = DataAPI.EquIndustryGet(industryVersionCD=u"010303",industry=u"",secID=u"",ticker=benchmark['consTickerSymbol'],intoDate=tradeDate,field=[u"ticker",u"industryName1"],pandas="1")
    BenchmarkIndustry = BenchmarkIndustry.where(BenchmarkIndustry.notnull(), 0)
    BenchmarkIndustry['weight'] = benchmark['weight']
    #计算基准指数的行业分配权重
    for industry in industryDF["industryName1"]:
        BenchmarkIndustrySum.ix[i,industry] = np.sum(BenchmarkIndustry[BenchmarkIndustry["industryName1"]==industry]["weight"])/np.sum(BenchmarkIndustry["weight"])
FundIndustrySum = FundIndustrySum.where(FundIndustrySum.notnull(), 0)
FundIndustryExposure = FundIndustrySum - BenchmarkIndustrySum
#plot
xLabel = FundIndustryExposure.columns.tolist()
for j in range(len(xLabel)):
    xLabel[j] = unicode(xLabel[j], "utf-8")

fig = plt.figure(figsize=(60,10))
font.set_size(10)
ax = fig.add_subplot(121)
ax = FundIndustryExposure.T.plot.bar(ax=ax)
ax.set_xticklabels(xLabel,rotation=0,fontproperties=font);
ax.set_ylabel(u'行业权重偏离',fontproperties=font);
```

最后输出的结果是公布的持仓数据相对于基准指数在各个行业上的权重偏离，挑几个显著的看一下，该基金在一季度末和四季度末是超配了银行股，而在二季度和三季度末是低配了银行股，且变动幅度非常之大，可见基金经理在银行的配置上是做了很干脆的交易（从最近的表现来看，超配的银行盈利不少呢，手哥不禁在心中暗暗的仰慕了一下这位基金经理）；食品饮料行业上可以看出基金经理在逐渐增加在该行业上的配置，所谓茅台不跌使人愁，看来就是这些公募基金经理买上去的；非银金融行业虽然一直在低配，但从幅度上来看，已经越来越小了，在2017年4季度末已经接近沪深300的水平。（回想起2018年1月份券商的行情，手哥不禁再次在心中暗暗的仰慕了一下这位基金经理，仰慕之余手哥也心生感慨，行情不会就是这帮人发动的吧）

```python
FundFactorsSum = pd.DataFrame(index = dateList,columns = [u"LCAP",u"CMRA",u"HSIGMA",u"VOL20",u"PE",u"PB"])
BenchmarkFactorsSum = pd.DataFrame(index = dateList,columns = [u"LCAP",u"CMRA",u"HSIGMA",u"VOL20",u"PE",u"PB"])
for i in range(len(dateList)):
    Date = dateList[i]
    #取持仓公布日前的最后一个交易日
    dateDF = DataAPI.TradeCalGet(exchangeCD=u"XSHG,XSHE",beginDate=Date.replace("-",""),endDate=Date.replace("-",""),field=[u"prevTradeDate",u"isOpen"],pandas="1")
    tradeDate = dateDF.iloc[0,0].replace("-","") if dateDF.iloc[0,1] == 0 else Date.replace("-","")
    #取得基金持仓
    FundHoldings = DataAPI.FundHoldingsGet(reportDate=u"",secID=u"",ticker=fundTicker,beginDate=u"",endDate=u"",secType="",field=u"",pandas="1")
    FundHoldingsQ = FundHoldings[FundHoldings['reportDate']==Date].sort_values(by="holdingTicker",axis = 0,ascending = True).reset_index(drop=True)
    #取得持仓股票的因子数据
    FundFactors = DataAPI.MktStockFactorsOneDayGet(tradeDate=tradeDate,secID=u"",ticker=FundHoldingsQ['holdingTicker'],field=[u"ticker",u"LCAP",u"CMRA",u"HSIGMA",u"VOL20",u"PE",u"PB"],pandas="1")
    FundFactors = FundFactors.where(FundFactors.notnull(), 0)
    FundFactors['weight'] = FundHoldingsQ['ratioInNa']
    #计算基金的因子值
    for theme in FundFactorsSum.columns:
        FundFactorsSum.ix[i,theme] = np.sum(FundFactors[theme]*FundFactors['weight'])/np.sum(FundFactors['weight'])
    #取得基准指数的成份股
    benchmark = DataAPI.mIdxCloseWeightGet(secID=u"",ticker=benchmarkTicker,beginDate=tradeDate,endDate=tradeDate,field=u"",pandas="1").sort_values(by="consTickerSymbol",axis = 0,ascending = True).reset_index(drop=True)
    #取得指数成份股的因子数据
    BenchmarkFactors = DataAPI.MktStockFactorsOneDayGet(tradeDate=tradeDate,secID=u"",ticker=benchmark['consTickerSymbol'],field=[u"ticker",u"LCAP",u"CMRA",u"HSIGMA",u"VOL20",u"PE",u"PB"],pandas="1")
    BenchmarkFactors = BenchmarkFactors.where(BenchmarkFactors.notnull(), 0)
    BenchmarkFactors['weight'] = benchmark['weight']
    #计算基准指数的因子值
    for theme in BenchmarkFactorsSum.columns:
        BenchmarkFactorsSum.ix[i,theme] = np.sum(BenchmarkFactors[theme]*BenchmarkFactors['weight'])/np.sum(BenchmarkFactors['weight'])
FundFactorsExposure = (FundFactorsSum - BenchmarkFactorsSum)/BenchmarkFactorsSum

#plot
fig = plt.figure(figsize=(40,10))
font.set_size(10)

ax = fig.add_subplot(121)
ax = FundFactorsExposure.T.plot.bar(ax=ax)
ax.set_xticklabels(FundFactorsExposure.columns,rotation=0,fontproperties=font);
ax.set_ylabel(u'风险因子暴露',fontproperties=font);
```

行业上做了简单分析之后，我们在看一下在风格上，基金组合跟基准指数有哪些差异。在这里，我们只简单的取6类风格因子的数据，市值（size）、动量（momentum）、波动率（volatility）、流动性（liquidity）、盈利（earnings）、价值（value），从优矿的初级因子库中，我们各选取一个因子作为一类风格因子的代表，对数市值LCAP、24月累计收益CMRA、历史波动HSIGMA、20日平均换手率VOL20、市盈率PE、市净率PB。每个因子的详细说明可以参考优矿数据说明文档。（手哥提示：优矿专业版可提供更为丰富的因子库，像此处的动量因子使用24个月的累计收益其实并不妥当，12月相对强势因子RSTR12实用性更好）

纵观整个2017年，LCAP，CMRA，PB因子带来的是正向收益，VOL20，PE带来的是负向收益，HSIGMA无明显方向的收益。（手哥提示：此处涉及到风格因子的收益归因，先按下不表）
从全年来看，该基金在LCAP、CMRA、HSIGMA和PE上没有持续稳定的暴露方向，VOL20、PB有明显的正向偏离，PE在三季度末有明显的负向偏离，但只持续了一段时间。其中，CMRA上半年的负向偏离给基金带来了负的超额收益，下半年的正向偏离带来了正的超额收益，这也能从一定程度上解释该基金在2017年上半年跑输基准而下半年跑赢基准，PB的正向偏离给基金带来了正的超额收益，PE在三季度短期的负向偏离带来了一定的正的超额收益，而VOL20全年的正向偏离带来的几乎都是负的超额收益，LCAP的差值较小，HSIGMA本身没有方向，这两个因子对基金组合并没有明显的超额贡献。
整体来看，该基金经理对风格的把握还是非常厉害了。（分析到这里，手哥不禁第三次在心中暗暗的仰慕了一下这位基金经理）
