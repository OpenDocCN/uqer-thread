# 回测数据保存监控 - observe使用示例

### 鉴于有UQERer使用回测框架时，比较关心回测数据记录，有必要重新提一下回测中的observe函数使用，请见下面策略示例

回测框架中的observe使用，见帮助文档9.1节
https://uqer.datayes.com/img/%E6%B7%B1%E5%85%A5%E6%B5%85%E5%87%BA%E9%87%8F%E5%8C%96%E5%AE%9E%E9%AA%8C%E5%AE%A4.pdf

#### observe保存的数据是回测输出 bt 的列

```python
## bt的最后几列就是我们保存的数据
bt = bt.set_index('tradeDate')
bt
```

#### observe保存的数据打印出来

```python
bt[[u'MA5-600680.XSHG', u'MA10-600680.XSHG', u'close-600680.XSHG']].plot(figsize=(10,6))
bt[[u'MA5-000025.XSHE', u'MA10-000025.XSHE', u'close-000025.XSHE']].plot(figsize=(10,6))
```

## 以上，可以看到observe可以提供另外一个debug工具
