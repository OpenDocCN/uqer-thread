# 涨停板与打板
所谓“打板”就是在一只股票涨停或者接近涨停时买入，投资的逻辑就是追涨，在涨停机制的限制下，涨停板的股票往往会在第二天或者后期有溢价效应。

量化的优点在于对历史数据的分析，在优矿的平台上，我们可以便捷的使用历史数据来分析一下打板及打板后的情况,以帮助我们理解和运用“打板”。

注意本帖所说的概率指的是频数的占比，而不是真正的概率分布（事实上这个没人知道是啥，过去假设股票服从集合布朗运动，现在也有从混沌系统去解释股票运动的）。根据大数定律，重复试验多次，随机事件发生的频率近似于它的概率。

## 一.涨停板次日表现分析

研究时间范围：2014.1.1-2016.6.30两年半   （时间选取太远对现在指导意义不大，但样本太少又没有说服力，所以就选取就近两年半）

研究股票范围:全A股

涨停板判断方式：收盘价等于涨停价（也就是我们没有考虑日内涨停后又跌下去的情况）

为了避免新股效应，判断当日涨停时，我们去除掉包括当日的前二十个交易日内上市的股票

```python
import pandas as pd
import numpy as np

#取交易日
from CAL.PyCAL import *
datedata=DataAPI.TradeCalGet(exchangeCD=u"XSHG",beginDate='20131201',endDate='20160630',field=['calendarDate','isOpen']) 
datedata=datedata[datedata['isOpen'] == 1]
date = map(lambda x: x[0:4]+x[5:7]+x[8:10], datedata['calendarDate'].values.tolist())    #多取一个月是为了判断是否在前二十个交易日上市

#取所有A股
stocks = DataAPI.EquGet(equTypeCD='A',listStatusCD='L',field='secID,nonrestfloatA')
universe = stocks['secID'].tolist()
```

```python
#取A股股票上市的数据
IPOinf=pd.DataFrame()      #用于存储处理后的股票ID和上市日期
IPO=DataAPI.EquIPOGet(eventProcessCD=u"",secID=universe,field=u"secID,listDate")
IPOinf[['secID','listDate']]=IPO[['secID','listDate']]             
IPO=IPO.set_index('secID')    #以股票ID为index                      
for i in range(len(universe)):
    IPOinf['listDate'][i]=int(filter(lambda x:x.isdigit(),IPO.ix[universe[i],0]))  #把上市日期处理成整型，方便以后调用处理
IPOinf=IPOinf[IPOinf['listDate']>20131201]                      #取在我们研究时间范围内的上市数据
IPOinf=IPOinf.sort(columns='listDate').set_index('listDate')          #以上市日期为index，方便查找在该日期上市的股票ID
```

准备工作做完了，下面可以调用API找出所有涨停板，以及涨停板后第二天的各价（最高，最低、开盘、收盘）情况

需要注意的是，这里判断是否涨停时使用的是未复权的收盘价，计算第二天各价的涨幅时，使用的是前复权股价。

由于股票较多，数据量比较大，我们先取每日的涨停板数据，做一个循环，取出两年半的涨停板数据（这里运行可能要花几分钟）

```python
#获取涨停板及涨停板第二日各价情况
uplimitinf=pd.DataFrame()   #用于存储所有涨停板及第二日数据
for i in range(len(date)-23):   #对日期进行循环
    
    delist=[]  #用于存储当日及前二十个交易日内上市的股票
    for j in range(20):
        if date[i+22-j] in IPOinf.index:
            delist=delist+IPOinf.ix[int(date[i+22-j]),0].values.tolist()
    universe=list(set(universe).difference(set(delist)))   #从全A股里剔除掉当日及前二十个交易日内上市的股票
    
    #判断当日涨停的股票，我这里使用的方法是判断当日收盘价是否等于涨停价
    data1=DataAPI.MktLimitGet(secID=universe,tradeDate=date[i+22],field=u"ticker,limitUpPrice,upLimitReachedTimes").set_index('ticker')
    data1=data1[data1['upLimitReachedTimes']>0]     #先剔除一部分当日没有涨停的股票
    data2=DataAPI.MktEqudGet(ticker=data1.index,tradeDate=date[i+22],field=u"ticker,closePrice").set_index('ticker')
    data3=pd.concat([data1,data2],axis=1)
    data3=data3[data3['closePrice'] == data3['limitUpPrice']]   #判断当日收盘价是否等于涨停价
    
    #调用当日涨停股票的前复权的股价信息，把第二日的各价信息存储到uplimitinf里，方便以后处理
    s=DataAPI.MktEqudAdjGet(tradeDate=date[i+23],ticker=data3.index,isOpen="1",field=u"ticker,tradeDate,preClosePrice,openPrice,highestPrice,lowestPrice,closePrice")
    uplimitinf=pd.concat([uplimitinf,s],axis=0)
```

我们取出的是涨停板次日的各价数据及前一日的收盘价，现在来看看取出的涨停板数据：

## 二.涨停板次日开盘涨幅对其最高价和收盘价涨幅的影响

现在我们来看一下不同涨停板次日开盘涨幅所对应的最高价涨幅的不同，希望对打板后在次日如何进行短线操作有一定指导作用。

我们不妨把将开盘个股分为：涨幅大于9.7%的（即开盘涨停）、涨幅为8%-9.7%、涨幅为6%-8%、涨幅为4%-6%、涨幅为2%-4%、涨幅为0-2%、涨幅-2%-0、以及开盘较大跌幅-2%-4%，和暴跌的（跌幅超过4%的）

易知，当开盘价涨幅高于某个值时，其最高价涨幅肯定也高于某个值。比如当开盘价涨幅大于5%时，那么最高价涨幅肯定在5%以上，所以我们分析一下开盘涨幅较小的情况

## 1.开盘涨幅在2%-4%时，最高价和收盘价涨幅的情况

```python
increase24=increase[(increase['increaseOpen']>=0.02 ) & (increase['increaseOpen']<=0.04)]
plt.hist(np.array(increase24['increaseHigh']),color=sns.desaturate("indianred", .8))
plt.title('increaseHigh')
```

直观上看，除了涨停股以外，最高价的涨幅均在2%以上，相当数量的股票还是涨停了，其它区间的分布比较均匀

```python
print 'Above 4%:'+str(float(len(increase24[increase24['increaseHigh']>0.04]))/len(increase24)), 'mean:'+str(increase24['increaseHigh'].mean())
print 'Quantile 50%:'+str(np.percentile(increase24['increaseHigh'],50))
```

当天最高价涨幅超过4%的概率为86.3%，意思就是说，我们可以设置在价格涨幅达到4%时，将股票卖掉，以大概率获取4%的收益。我们有一半的概率获取8%的收益。但如果没有按照我们的预想在4%出掉货，不得不持有至收盘，那尾盘的股价情况如何呢,我们看下图：

```python
plt.hist(np.array(increase24['increaseClose']),color=sns.desaturate("indianred", .8))
plt.title('increaseClose')
```

```python
print 'Above 0:'+str(float(len(increase24[increase24['increaseClose']>0]))/len(increase24)), 'mean:'+str(increase24['increaseClose'].mean())
print 'Above 4%:'+str(float(len(increase24[increase24['increaseClose']>0.04]))/len(increase24))
```

到了尾盘的时候，我们能够获益的概率为75.6%，但能获取4%的收益的概率只有47.89%了，不到一半的概率。这时候收盘涨停的概率还是比较大的

## 2.开盘涨幅在0-2%时，最高价和收盘价涨幅的情况

```python
increase02=increase[(increase['increaseOpen']>=0 ) & (increase['increaseOpen']<=0.02)]
plt.hist(np.array(increase02['increaseHigh']),color=sns.desaturate("indianred", .8))
plt.title('increaseHigh')
print 'Above 0:'+str(float(len(increase02[increase02['increaseHigh']>0]))/len(increase02)), 'mean:'+str(increase02['increaseHigh'].mean())
print 'Above 4%:'+str(float(len(increase02[increase02['increaseHigh']>0.04]))/len(increase02))
```

我们可以看到在开盘涨幅只有0-2%时，最高价能以99.1%的概率有正的涨幅，涨幅达到4%的概率为67.4%，比在开盘涨幅2%-4%时要低一些，也就是说，开盘涨幅低一些时，我们对股价的期待要低一些，设置卖出的价格也可以相应低一些。再来看看尾盘的情况：

```python
increase02=increase[(increase['increaseOpen']>=0 ) & (increase['increaseOpen']<=0.02)]
plt.hist(np.array(increase02['increaseClose']),color=sns.desaturate("indianred", .8))
plt.title('increaseClose')
print 'Above 0:'+str(float(len(increase02[increase02['increaseClose']>0]))/len(increase02)), 'mean:'+str(increase02['increaseClose'].mean())
print 'Above 4%:'+str(float(len(increase02[increase02['increaseClose']>0.04]))/len(increase02))
```

从上图可以看出，当开盘价涨幅在0-2%，其收盘涨停的概率明显下降，跌停的概率上升，剩下的基本上呈现正态分布。记过计算收盘为红的概率为61.5%,胜率明显下降。所以当你持有的昨日涨停板的股票今天开盘在0-2%之间时，就得做好准备,有67.4%的概率能够以4%卖出，不然到尾盘就只剩下61.5%的机会能盈利了

## 3.开盘涨幅在-2%-0时，最高价和收盘价涨幅的情况

```python
increase20=increase[(increase['increaseOpen']>= -0.02 ) & (increase['increaseOpen']<=0)]
plt.hist(np.array(increase20['increaseHigh']),color=sns.desaturate("indianred", .8))
plt.title('increaseHigh')
print 'Above 0:'+str(float(len(increase20[increase20['increaseHigh']>0]))/len(increase20)), 'mean:'+str(increase20['increaseHigh'].mean())
print 'Above 4%:'+str(float(len(increase20[increase20['increaseHigh']>0.04]))/len(increase20))
```

能够以4%的涨幅卖出的概率就变得更低了为48.9%，意思是昨日涨停板上拿的票，明天以4%涨幅委托卖出的概率为48.9%。这时候再以收益4%的委托价卖出，是非常不明智的

我们如果仅仅以不亏损为前提，意思是以昨日买入价委托卖出，基本上有84.5%的概率不亏本。再来看看尾盘的情况：

```python
increase20=increase[(increase['increaseOpen']>= -0.02 ) & (increase['increaseOpen']<=0)]
plt.hist(np.array(increase20['increaseClose']),color=sns.desaturate("indianred", .8))
plt.title('increaseClose')
print 'Above 0:'+str(float(len(increase20[increase20['increaseClose']>0]))/len(increase20)), 'mean:'+str(increase20['increaseClose'].mean())
print 'Above 4%:'+str(float(len(increase20[increase20['increaseClose']>0.04]))/len(increase20))
```

只剩下45.8%的概率能够以不亏收尾，平均涨幅已经变为负了，这时候建议该日应该提前以一个收益低的委托价卖出，比如2%

```python
print 'Above 1%:'+str(float(len(increase20[increase20['increaseHigh']>0.01]))/len(increase20))
```

可以看到，如果以涨幅2%的委托价卖出，有75.7%的概率能够卖出

```python
#定义为函数，方便获取最高价涨幅和收盘价涨幅所在的各个百分比的频率，存为list
def get_HighP(x):
    result=[]
    result=result+[float(len(x[x['increaseHigh']>0.097]))/len(x)]
    for i in range(5):
        result=result+[float(len(x[x['increaseHigh']>0.01*(4-i*2)]))/len(x)]
    return result
def get_CloseP(x):
    result=[]
    result=result+[float(len(x[x['increaseClose']>0.097]))/len(x)]
    for i in range(5):
        result=result+[float(len(x[x['increaseClose']>0.01*(4-i*2)]))/len(x)]
    return result
```

```python
#得到各种条件下的各价涨幅的概率
def get_probility(x):
    #从涨停板次日各价数据得到次日各价涨幅，存在新的DataFrame里
    increase=pd.DataFrame()
    increase['increaseOpen']=x['openPrice']/x['preClosePrice']-1
    increase['increaseHigh']=x['highestPrice']/x['preClosePrice']-1
    increase['increaseClose']=x['closePrice']/x['preClosePrice']-1
    result=pd.DataFrame()
    #以开盘价的涨幅为筛选条件
    increase24=increase[(increase['increaseOpen']>=0.02 ) & (increase['increaseOpen']<=0.04)]
    increase02=increase[(increase['increaseOpen']>=0 ) & (increase['increaseOpen']<=0.02)]
    increase20=increase[(increase['increaseOpen']>= -0.02 ) & (increase['increaseOpen']<=0)]
    increase42=increase[(increase['increaseOpen']>= -0.04 ) & (increase['increaseOpen']<=-0.02)]
    increase4=increase[increase['increaseOpen']<=-0.04]
    #得到各个开盘涨幅条件下最高价和收盘价的涨幅情况
    result['开盘价涨幅:2%-4%']=get_HighP(increase24)+get_CloseP(increase24)
    result['0-2%']=get_HighP(increase02)+get_CloseP(increase02)
    result['-2%-0']=get_HighP(increase20)+get_CloseP(increase20)
    result['-4%-(-2%)']=get_HighP(increase42)+get_CloseP(increase42)
    result['<-4%']=get_HighP(increase4)+get_CloseP(increase4)
    result['不论开盘']=get_HighP(increase)+get_CloseP(increase)
    result.index=['最高价涨停','最高价涨幅:>4%','最高价涨幅:>2%','最高价涨幅:>0%','最高价涨幅:>-2%','最高价涨幅:>-4%','收盘涨停','收盘价涨幅:>4%','收盘价涨幅:>2%','收盘价涨幅:>0%','收盘价涨幅:>-2%','收盘价涨幅:>-4%']
    return result
```

上述DataFrame里列出了之前的各个涨幅的条件概率，读者也可以修改函数，以达到自己研究的目的。

# 三.板次数对次日收益的影响
我们依然沿用第二部分的分析方法，看在连板次数不同的情况下，次日低开高开平开对涨停板次日最高价和收盘价的影响。
## 1.一个涨停板(当日涨停，前一日没有涨停)。

```python
#先只选取tradedate和ticker的数据方便判断板的次数
tradedateinf=uplimitinf[[1,2]]
tradedate=tradedateinf.drop_duplicates(subset='tradeDate')
tradedate=tradedate['tradeDate'].values.tolist()
tradedateinf.set_index('tradeDate',inplace=True)
```

```python
#选取只有一个连板的涨停板次日各价数据
uplimitsOne=pd.DataFrame()  #用于存储一连板的数据
for i in range(len(tradedate)-1):   #依然对交易日作循环
    a=tradedateinf.loc[tradedate[i]]   #前一日涨停股票
    b=tradedateinf.loc[tradedate[i+1]]  #当日涨停股票
    if len(a) == 1:
        a=a.tolist()
    else:
        a=a['ticker'].tolist()
    if len(b) == 1:
        b=b.tolist()
    else:
        b=b['ticker'].tolist()
    tickers=list(set(b).difference(set(a)))   #当日涨停股票去掉前一日已经涨停的股票
    tmp=uplimitinf[(uplimitinf['tradeDate'] == tradedate[i+1]) & (uplimitinf['ticker'].isin(tickers))]   #把刚才筛选后的涨停板数据存在uplimitsOne里
    uplimitsOne=pd.concat([uplimitsOne,tmp],axis=0)
```

```python
print '一连板个数：'+str(len(uplimitsOne))
```

```python
get_probility(uplimitsOne)
```

我们可以和之前的结果进行对比，发现得到高收益的概率均下降了。开盘涨幅即在2%-4%之间的个股，当天最高价超过4%概率为86.0%，比之前降低了一些。我们依然可以在这种情况下设置一个4%的委托，获取4%的收益。开盘涨幅即在-4%以下时，我们获取正的收益的概率为62.8%，比之前提高了一些。

对于尾盘而言，获取高收益的概率也下降了，但是跌幅大的概率也下降了。收盘再涨停的概率比整体上要低。开盘涨幅即在2%-4%之间的个股，收盘价涨幅大于-4%的概率为94.8%,比之前高一些。但在开盘涨幅较低的情况下，还是不建议持有到尾盘，尽量冲高抛售。

## 2.二连板(当日涨停，前一日涨停，再前一日未涨停)。

```python
#选取二连板的涨停板次日各价数据
uplimitsTwo=pd.DataFrame()    #用于存储二连板的数据
for i in range(len(tradedate)-2):
    a=tradedateinf.loc[tradedate[i]]   #前两日涨停股票
    b=tradedateinf.loc[tradedate[i+1]]  #前一日涨停股票
    c=tradedateinf.loc[tradedate[i+2]]  #当日涨停股票
    #存为list
    if len(a) == 1:
        a=a.tolist()
    else:
        a=a['ticker'].tolist()   
    if len(b) == 1:
        b=b.tolist()
    else:
        b=b['ticker'].tolist()
    if len(c) == 1:
        c=c.tolist()
    else:
        c=c['ticker'].tolist()
    tickers=list(set(c).intersection(set(b)))    #前一日涨停股票与当日涨停股票取交集
    tickers=list(set(tickers).difference(set(a)))  #去掉前两日已经涨停的股票
    tmp=uplimitinf[(uplimitinf['tradeDate'] == tradedate[i+2]) & (uplimitinf['ticker'].isin(tickers))]
    uplimitsTwo=pd.concat([uplimitsTwo,tmp],axis=0)
```

```python
print '二连板个数：'+str(len(uplimitsTwo))
```

```python
get_probility(uplimitsTwo)
```

二连板当开盘价涨幅为2%-4%时，最高价涨幅超过4%的概率为85.5%；当开盘低开时，最高价涨幅〉4%的概率就小于50%了，但仍有超过一半的概率获取正的收益。收盘结果如下：

当开盘价涨幅为2%-4%时,收盘涨停的概率有31%，比一连板和整体上都要高一些，但能获取正的收益的概率比一连板时降低了。

```python
#选取三连板的涨停板次日各价数据
uplimitsThree=pd.DataFrame()
for i in range(len(tradedate)-3):
    a=tradedateinf.loc[tradedate[i]]  #前三日涨停股票
    b=tradedateinf.loc[tradedate[i+1]]  #前两日涨停股票
    c=tradedateinf.loc[tradedate[i+2]]  #前一日涨停股票
    d=tradedateinf.loc[tradedate[i+3]]  #当日涨停股票
    if len(a) == 1:
        a=a.tolist()
    else:
        a=a['ticker'].tolist()
    if len(b) == 1:
        b=b.tolist()
    else:
        b=b['ticker'].tolist()
    if len(c) == 1:
        c=c.tolist()
    else:
        c=c['ticker'].tolist()
    if len(d) == 1:
        d=d.tolist()
    else:
        d=d['ticker'].tolist()
    tickers=list(set(d).intersection(set(c)))   #前一日涨停股票与当日涨停股票取交集
    tickers=list(set(tickers).intersection(set(b)))  #与前两日涨停股票取交集
    tickers=list(set(tickers).difference(set(a)))  #去掉前三日涨停股票
    tmp=uplimitinf[(uplimitinf['tradeDate'] == tradedate[i+3]) & (uplimitinf['ticker'].isin(tickers))]
    uplimitsThree=pd.concat([uplimitsThree,tmp],axis=0)
```
