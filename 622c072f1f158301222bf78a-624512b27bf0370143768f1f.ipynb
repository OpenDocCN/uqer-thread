使用朴素贝叶斯模型来预测，首先来学习下这个模型



````python
import numpy as np
rng = np.random.RandomState(1)
````

````python
# 生成模拟数据矩阵，6*100，有6个分类，每个分类100个特征
X = rng.randint(5, size=(6, 100)) 
X
````

````python
Y = np.array([1, 2, 3, 4, 4, 5]) # Y 代表不同类别的标签，1，2,3,44,5
````

````python
from sklearn.naive_bayes import BernoulliNB
clf = BernoulliNB()
clf.fit(X, Y) # 训练模型
````

````python
# 没有测试集，所以还是从训练集中取出一个类别来看看分类是否准确
x = X[2]
clf.predict(x) # 预测该类别
````
