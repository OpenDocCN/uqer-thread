```python
import numpy as np
import pandas as pd
from pandas import DataFrame

import matplotlib.pyplot as plt

from collections import defaultdict
```

```python
#策略数据准备
# retrieves the historical prices for a list of stocks
def all_dataPre(tickers, start, end):       
    all_data = {} #record the return data; key:stock_id; value: factor value in trading_days (in Pandas)
    def data(ticker):
        #--装载全部数据
        ticker_data = DataFrame()
        ticker_data = DataAPI.MktEqudGet(secID=ticker,beginDate=start,endDate=end,field=u"tradeDate,openPrice,closePrice,turnoverVol",pandas="1")
        ticker_data =  ticker_data.rename(columns={'tradeDate':'DATE','openPrice':'OPEN','closePrice':'CLOSE','turnoverVol':'VOLUMN'})
        ticker_data['DATE'] =  pd.to_datetime( ticker_data['DATE']) #因为DataAPI所有日期都是以字符串形式存储的，因此此处需要转换成日期时间型,这里很重要！
        ticker_data =  ticker_data.set_index('DATE')#转换后index是DatetimeIndex型
        return ticker_data   
    for ticker in tickers:
         all_data[ticker] = data(ticker)
    return all_data
```

```python
#计算策略信号
def calc_mom(signal, lookback, lag):
    mom_ret = signal.shift(lag).pct_change(lookback)
    ranks = mom_ret.rank(axis=1, ascending=False)
    demeaned = ranks.sub(ranks.mean(axis=1),axis='index')
    return demeaned.div(demeaned.std(axis=1),axis='index')
```

```python
compound = lambda x : (1 + x).prod() - 1
daily_sr = lambda x: x.mean() / x.std()
```

```python
#组合收益
def port_returns(prices,signals,lb, hold):       
     # Compute portfolio weights
    freq = '%dB' % hold
    port = calc_mom(signals, lb,lag=1)
    daily_rets = prices.pct_change()
  
    # Compute portfolio returns
    #the old syntax
    port = port.shift(1).resample(freq,how='first')
    returns = daily_rets.resample(freq,how=compound) 
    #the new syntax
    # port = port.shift(1).resample(freq).first()
    # returns = daily_rets.resample(freq).apply(compound)         
    return (port * returns).sum(axis=1)

#可视化
def heatmap(df, cmap=plt.cm.gray_r):
    fig = plt.figure()
    ax = fig.add_subplot(111)
    axim = ax.imshow(df.values, cmap=cmap, interpolation='nearest')
    ax.set_xlabel(df.columns.name)
    ax.set_xticks(np.arange(len(df.columns)))
    ax.set_xticklabels(list(df.columns))
    ax.set_ylabel(df.index.name)
    ax.set_yticks(np.arange(len(df.index)))
    ax.set_yticklabels(list(df.index))
    plt.colorbar(axim)
```

```python
#回测区间
start = '20140104'                       # 回测起始时间
end   = '20160414'                         # 回测结束时间

#投资域：自定，但需去除ST股；去除流动性差的股票；去除新上市或复牌的股票；剔除了当天停牌、退市和数据异常证券。
universe = ['000156.XSHE','000503.XSHE','000504.XSHE',]

#获取回测期间全部数据
all_data = all_dataPre(universe, start, end)

#获取计算收益原始数据
prices = DataFrame({tic:data['CLOSE'] for tic,data in all_data.iteritems()})
#准备信号策略signals ，
signals =  prices
```

```python
lookbacks = range(20, 90, 5)
holdings = range(20, 90, 5)

dd = defaultdict(dict)
srMax = 1.0
srlb  = 20
srhold= 20

#策略暴力选优
for lb in lookbacks:
    for hold in holdings:
        #generate strategy sharp ratio
        rets = port_returns(prices,signals,lb, hold)
        dd[lb][hold] =  daily_sr(rets) * np.sqrt(252 / hold)
        if dd[lb][hold]>srMax:
            srMax = dd[lb][hold]
            srlb  = lb
            srhold= hold
ddf = DataFrame(dd)
ddf.index.name = 'Holding Period'
ddf.columns.name = 'Lookback Period'
heatmap(ddf)
```
