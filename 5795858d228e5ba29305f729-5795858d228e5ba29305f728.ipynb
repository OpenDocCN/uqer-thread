## ** 1，引言 **
之前发的异动事件研究，可能较为简单，研究方面也不够全面，这里想对根据[1]中介绍的思路，异动事件做一个更为详细深入的分析。

在通常情况下，股票与指数的日内走势是随波逐流的关系。但是，在某些特殊的交易日里，股票可能在盘中频繁出现与指数走势背道而驰的情况。这种个性十足的价格异动，我们称之为“特立独行”事件。本文主要研究这类特立独行事件的特性，并据此制订相关策略。

## **3，异动事件的量化识别**
[1]认为，就异动事件如何进行量化识别的问题，相关性可能是一个很好的切入点：在“随波逐流”的常态下，个股与指数同步涨跌，两者分钟收益率序列的相关度也会比较高；而在“特立独行”的情形下，由于个股与指数频繁出现涨跌背离，可以预期，相关度将显著降低。<br>
因此，本文继续以水晶广电为研究目标，给出了11月16日前后每个交易日水晶光电与上证综指分钟收益序列的皮尔逊相关系数。正如我们所预期的，在大多数交易日里，相关系数的数值在一个较高的数值水平（约 0.4）附近徘徊，而恰恰在 11 月 16日这一天，相关系数出现了异常的低值（低于0.05）。<br>
顺着这一思路，我们可以考虑将每日的相关系数作为筛选“特立独行”事件的因子。筛选的最简单做法是，给定一个绝对阈值（比如 0.1），将相关系数小于该阈值的交易日认定为“特立独行”。然而，这种绝对阈值的做法存在严重的缺陷：不同股票与上证综指的相关度水平天然不同，统一阈值无法适用。比如，0.1 的绝对阈值，对于大市值的股票而言可能偏低（很少选出样本），对于小市值的股票则可能偏高（频繁选出样本）。因此，为避免此类样本偏差，客观的筛选方案应将每只股票相关系数的历史水平考虑在内。

```python
coe=[]
date_list=[]
for i in range(28):
    index_list=[]
    stock_list=[]
    index=DataAPI.MktBarHistOneDayGet(securityID='000001.XSHG',date=(datetime.datetime(2015,10,23)+datetime.timedelta(days=i)).strftime('%Y%m%d')).closePrice
    stock=DataAPI.MktBarHistOneDayGet(securityID='002273.XSHE',date=(datetime.datetime(2015,10,23)+datetime.timedelta(days=i)).strftime('%Y%m%d')).closePrice
    for j in range(len(index)-1):
        index_list.append(np.log(index[j+1]/index[j]))
        stock_list.append(np.log(stock[j+1]/stock[j]))
    index_price=Series(index_list)
    stock_price=Series(stock_list)
    if np.isnan((np.mean(index_price*stock_price)-np.mean(index_price)*np.mean(stock_price))/(np.std(index_price)*np.std(stock_price))):
        continue
    else:
        coe.append((np.mean(index_price*stock_price)-np.mean(index_price)*np.mean(stock_price))/(np.std(index_price)*np.std(stock_price)))
        date_list.append((datetime.datetime(2015,10,23)+datetime.timedelta(days=i)).strftime('%Y%m%d'))
print coe
fig = plt.figure(figsize=(16,6))
ax = fig.add_subplot(111)
fig.set_tight_layout(True)
ax.grid(axis=u"y")
plt.bar(range(19),coe,width=0.5)
plt.xticks(range(19), (date_list))  
plt.title(u"日内相关系数：水晶广电&上证指数",fontproperties=font, fontsize=16)
plt.show()
```

综上讨论，我们给出“特立独行”异动股票的筛选方案，步骤如下：
>1,取交易日 t、股票 s 的日内分钟收益率序列，计算其与上证综指当日分钟收益率序列的皮尔逊相关系数 rho(s,t)；
> <br>
>2,对于每只股票 s，将每日的相关系数 rho(s,t) 沿纵向标准化，得到相关度因子 f(s,t)；
> <br>
>3,将相关度因子 f(s,t) 低于阈值λ，视为发生“特立独行”异动事件。
<br>

其中，步骤（2）中“纵向标准化”的具体做法为：将当日相关系数 rho(s,t)减去相关系数过去 10 个交易日的均值，再除以 10 个交易日的标准差。步骤（3）中阈值λ取 -3。筛选方案的直观意义是，选出相关系数处于历史平均水平 3 倍标准差以下的交易日。
<br>

按照以上筛选方案，除水晶光电（2015-11-16）之外，还可以选出一个广为人知的案例——梅雁吉祥（600868.SH）。梅雁吉祥在 2015 年 8 月 6 日触发了“特立独行”事件，而两个交易日之前正是其“王的女人”故事的起点——梅雁吉祥 8 月4 日的临时公告称“证金公司已成为第一大股东”。如我们所熟知的，梅雁吉祥在 8 月 6 日之后连续录得 7 个涨停，超额收益异常可观。

## **4,实证研究**
在接下来的部分中我们将对异动事件作相关的实证研究，考虑到在线平台的运算速度有限，本文仅选取了沪深300作为股票池，时间跨度为2015年1年，历史回测筛选出的“特立独行”异动样本总共 7351 次，涉及股票 2530 只，分布在 640 个交易日。

```python
universe=set_universe('HS300')
Lambda=-3
dif_stock=[]
event_date=[]
event=[]
i=0
p=0
for stk in universe:
    p=p+1
    print p
    for j in range(365):
        rho_list=[]
        k=0
        i=0
        while(i<10):    #前20个交易日相关系数
            index_list=[]
            stock_list=[]
            index=DataAPI.MktBarHistOneDayGet(securityID='000001.XSHG',date=(datetime.datetime(2015,1,1)+datetime.timedelta(days=j)-datetime.timedelta(days=k+1)).strftime('%Y%m%d')).closePrice
            stock=DataAPI.MktBarHistOneDayGet(securityID=stk,date=(datetime.datetime(2015,1,1)+datetime.timedelta(days=j)-datetime.timedelta(days=k+1)).strftime('%Y%m%d')).closePrice
            for kk in range(len(index)-1):
                index_list.append(np.log(index[kk+1]/index[kk]))
            for kk in range(len(stock)-1):
                stock_list.append(np.log(stock[kk+1]/stock[kk]))
            index_price=Series(index_list)
            stock_price=Series(stock_list)
            k=k+1
            if k>14:
                break
            if np.isnan((np.mean(index_price*stock_price)-np.mean(index_price)*np.mean(stock_price))/(np.std(index_price)*np.std(stock_price))):
                continue
            else:
                rho_list.append((np.mean(index_price*stock_price)-np.mean(index_price)*np.mean(stock_price))/(np.std(index_price)*np.std(stock_price)))
                i=i+1
        indexs_list=[]   #当日相关系数
        stocks_list=[]
        indexs=DataAPI.MktBarHistOneDayGet(securityID='000001.XSHG',date=(datetime.datetime(2015,1,1)+datetime.timedelta(days=j)).strftime('%Y%m%d')).closePrice
        stocks=DataAPI.MktBarHistOneDayGet(securityID=stk,date=(datetime.datetime(2015,1,1)+datetime.timedelta(days=j)).strftime('%Y%m%d')).closePrice
        for jj in range(len(indexs)-1):
            indexs_list.append(np.log(indexs[jj+1]/indexs[jj]))
        for jj in range(len(stocks)-1):
            stocks_list.append(np.log(stocks[jj+1]/stocks[jj]))
        indexs_price=Series(indexs_list)
        stocks_price=Series(stocks_list)
        if np.isnan((np.mean(indexs_price*stocks_price)-np.mean(indexs_price)*np.mean(stocks_price))/(np.std(indexs_price)*np.std(stocks_price))):
            continue
        else:
            rho=((np.mean(indexs_price*stocks_price)-np.mean(indexs_price)*np.mean(stocks_price))/(np.std(indexs_price)*np.std(stocks_price)))
            f=(rho-np.mean(rho_list))/np.std(rho_list)
            if f<Lambda:
                dif_stock.append(stk)
                event_date.append((datetime.datetime(2015,1,1)+datetime.timedelta(days=j)))
                event.append([stk,(datetime.datetime(2015,1,1)+datetime.timedelta(days=j))])
```

为了考察“特立独行”异动的事件收益，我们将“特立独行”事件的触发日记为 T+0 日，从 T+1 日开盘买入全体异动样本，持仓 60 个交易日至 T+60 日收盘。 下图给出了超额收益的测算结果。其中，红柱为全体样本相对于上证指数的平均逐日超额收益（右轴），蓝线为平均累积超额收益（左轴）。

根据上文所做的研究，尝试开发相应的股票交易策略。买入列表按照如下步骤选取：

>（1）以每日“特立独行”异动样本为原始股票池；
> <br>
>（2）选取其中属于“逆势涨”的股票；
> <br>
>（3）剔除其中近期涨幅过高的股票。
<br>

其中，“逆势涨”以“当日个股超额收益大于 0”为近似判断；“近期涨幅过高”以“过去 10 个交易日累积收益大于20%”为判断标准。
<br>
投资组合的资金管理按照如下规则进行：
>（1）将初始资金等分为 5 份（或称 5 个资金通道）；
>（2）每日收盘后选出合格的股票标的，次日以开盘价买入，等权占用其中一个通道的资金，持有 40 个交易日后以收盘价卖出；（根据超额收益图在30个交易日时收益较高，但是空仓率也相对高，综合考虑取40个交易日。）
>（3）当所有资金通道已全被占用时，若出现新的标的，则卖出最早通道中的股票，所得资金用于买入新的标的股票。
