## **阻力位和支撑位的计算**
#### 阻力位
> - 阻力位1 (R1) = 上月度收盘价 + (上月度最高价-上月度最低价） × 1.1 / 12 
- 阻力位2 (R2) = 上月度收盘价 + (上月度最高价-上月度最低价） × 1.1 / 6 
- 阻力位3 (R3) = 上月度收盘价 + (上月度最高价-上月度最低价） × 1.1 / 4 
- 阻力位4 (R4) = 上月度收盘价 + (上月度最高价-上月度最低价） × 1.1 / 2 
- 阻力位5 (R5) = R4 + (R4 - R3) × 1.168 
- 阻力位6 (R6) = 上月度最高价 / 上月度最低价 × 上月度收盘价 

#### 支撑位
> - 支撑位1 (S1) = 上月度收盘价 - (上月度最高价-上月度最低价） × 1.1 / 12 
- 支撑位2 (S2) = 上月度收盘价 - (上月度最高价-上月度最低价） × 1.1 / 6
- 支撑位3 (S3) = 上月度收盘价 - (上月度最高价-上月度最低价） × 1.1 / 4
- 支撑位4 (S4) = 上月度收盘价 - (上月度最高价-上月度最低价） × 1.1 / 2
- 支撑位5 (S5) = S4 - (S3 - S4) × 1.168
- 支撑位6 (S6) = 上月度收盘价 - (R6 - 上月度收盘价)

### 反转交易
>IF 本月开盘价>支撑位3(S3) 且 本月开盘价<阻力位3(R3)
当价格跌到S3时做多。止盈方案为R1卖出40%﹑R2卖出30%与R3卖出30%，如突破R1后未到R2又回落到R1，则卖出剩余全部；
如突破R2后未到R3，又回到R2，同样卖出剩余全部；止损方案为跌至S4；

#### 当价格涨到R3时做空。止盈点分别为S1﹑S2和S3，止损点为R4

```python
from CAL.PyCAL import *
from pandas import DataFrame, Series
import pandas as pd
import numpy as pd
from datetime import datetime
def get_day_range_last_month(account):
    cal = Calendar('China.SSE')
    referenceDate = account.current_date

    period = Period('-'+str(referenceDate.day)+'D')
    firstDay = Date(referenceDate.year, referenceDate.month,1)
    firstDayOfPrevMonth = cal.advanceDate(firstDay, Period('-1M'), BizDayConvention.Following)
    lastDayOfPrevMonth = cal.advanceDate(referenceDate, period, BizDayConvention.Preceding)
    
    return firstDayOfPrevMonth.strftime('%Y%m%d'), lastDayOfPrevMonth.strftime('%Y%m%d')

def Camarilla(hist_max, hist_min, last_close):
    hist_volatile = hist_max - hist_min
    R1 = last_close + hist_volatile * 1.1 /12.0
    R2 = last_close + hist_volatile * 1.1 /6.0
    R3 = last_close + hist_volatile * 1.1 /4.0
    R4 = last_close + hist_volatile * 1.1 /2.0
    R5 = R4 + (R4-R3)*1.168 
    R6 = hist_max/hist_min * last_close

    S1 = last_close - hist_volatile * 1.1 /12.0
    S2 = last_close - hist_volatile * 1.1 /6.0
    S3 = last_close - hist_volatile * 1.1 /4.0
    S4 = last_close - hist_volatile * 1.1 /2.0
    S5 = S4 - (S4-S3)*1.168 
    S6 = last_close - (R6 - last_close)
    
    return R1,R2,R3,R4,R5,R6,S1,S2,S3,S4,S5,S6

#shgcy大神的止损方法
def stop_loss(account):
    if should_clear(account):
        for stock in account.avail_secpos.keys():
            order_to(stock, 0)
        return True
    return False
    

#shgcy大神的止损方法,大盘20日均线拐头向下清仓
def should_clear(account):
    cal = Calendar('China.SSE')
    today = account.current_date
    prev20Day = cal.advanceDate(today, Period('-21B'), BizDayConvention.Preceding)
    
    n = 20
    sz_hist = DataAPI.MktIdxdGet(ticker='399001', beginDate=prev20Day.strftime('%Y%m%d'),field=['tradeDate', 'closeIndex'], pandas='1')
    yesterday_ma = sz_hist[0:n].mean()
    today_ma = sz_hist[1:n+1].mean()
    # print 'today_ma.closeIndex - yesterday_ma.closeIndex', str(today_ma.closeIndex - yesterday_ma.closeIndex)
    return today_ma.closeIndex < yesterday_ma.closeIndex


    
```

### **来自@liniu大大的选股和估值方法**

```python
def preceding_date(date):
    cal = DataAPI.TradeCalGet(exchangeCD=u"XSHG",beginDate='20110101',endDate=date,field=['calendarDate','isOpen'],pandas="1")
    cal = cal[cal['isOpen']==1]
    date = cal['calendarDate'].values[-2].replace('-','')
    return date

def spreadRateByIntrinsicValue(account, overflow=0.0, precedingDate=True):
    stock_list = account.universe
    current_date = account.current_date
    date = current_date.strftime('%Y%m%d')
    if precedingDate:
        date = preceding_date(date)
    eq_EPS_EGRO = DataAPI.MktStockFactorsOneDayGet(tradeDate=date,secID=stock_list,field=['secID','EPS','EGRO'],pandas="1")
    eq_EPS_EGRO['Value'] = eq_EPS_EGRO['EPS']*(8.5+2*eq_EPS_EGRO['EGRO']/5)
    eq_EPS_EGRO = eq_EPS_EGRO.dropna()
    spread_rate = []
    for stock, intrinsic_value in zip(eq_EPS_EGRO['secID'].values, eq_EPS_EGRO['Value'].values):
        intrinsic_value = intrinsic_value*(1+overflow)
        reference_price = account.referencePrice[stock]
        if reference_price > 0 and reference_price < intrinsic_value:
            spread_rate.append((stock, (intrinsic_value-reference_price)/reference_price))
    return sorted(spread_rate, key=lambda k: k[-1], reverse=True)
```
