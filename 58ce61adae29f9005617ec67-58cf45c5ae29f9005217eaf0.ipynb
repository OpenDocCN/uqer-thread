# tensorflow12 单机Optimizer选择

![](http://storage-uqer.datayes.com/57b19ec7228e5b79a7759010/5dd19284-0d19-11e7-bdeb-0242ac140004)
最近在看训练的一些bolg和paper简单的对训练求解总结一下。

1 对时刻t的参数使用相同的学习率$\eta$

如果估计模型稀疏很小，可以优先使用Nesterov。

tf.train.MomentumOptimizer(use_nesterov=True)

2 对时刻t的参数使用不同的的学习率$\eta$，也就是每个权重参数$\omega_i$对应一个学习率$\eta_{i,t}$

如果估计有一定的稀疏或者使用dropout防止过拟合可以优先使用Adam或者Nadam。

tf.train.AdamOptimizer
