# BOLL斜率扭转买卖法

布林线斜率策略
当布林线中轨斜率小于0，通道向下，同时布林线中轨下轨斜率趋近于0.01，买入股票
当布林线中轨斜率大于0，通道向上，同时布林线中轨上轨斜率趋近于0.01，卖出股票

价格单位长度=20天内的（最高价-最低价）/20
实时Boll斜率公式=(今天boll值-昨天boll值)/价格单位长度

这里用格力电器从2006年到今天的数据的来回测，以上证指数为基准。
年化收益率达到了34%，不过回撤也达到了65%，都是在牛熊转换的时候出现

不过实际使用时我会结合级别突破体系，比如日线突破周级别的boll线等等，实际回撤不会这么大，比如我2018年的回撤只有5%
