```python
import pandas as pd
import numpy as np


# 资源占用情况
def compareByOccupation(list_ticker, index = '资源占用'):
    df = DataAPI.FdmtBSAllLatestGet(ticker=list_ticker,secID=u"",reportType=u"A",endDate=u"",beginDate=u"",year="",field=u"",pandas="1")
    df_result = 1
    
    for i in df:
        df[i] = np.nan_to_num(df[i])
    
    dict_stock = {}
    for ticker in df.groupby('ticker'):
        df_temp = ticker[1][:10]
        stockcode =  ticker[0]
        stockname =  df_temp.secShortName.tolist()[0]
        dict_stock[stockcode] = stockname
        
        
        df_temp['occupation'] = ((df_temp['AP']) + (df_temp['advanceReceipts']) + (df_temp['othPayable'])) -  ((df_temp['AR']) + (df_temp['prepayment']) + (df_temp['othReceiv'])) 
        df_temp['assets'] = ((df_temp['TShEquity'])+ (df_temp['STBorr'])+ (df_temp['NCLWithin1Y'])+ (df_temp['LTBorr'])+ (df_temp['bondPayable'])+ (df_temp['LTPayable']))
        
        if index == '资源占用':
            ratio = 'occupation'
        if index == '应付账款':
            ratio = 'AP'
        if index == '预收款项':
            ratio = 'advanceReceipts'
        if index == '其他应付款':
            ratio = 'othPayable'
            
            
        df_temp[ratio] = np.round(df_temp[ratio] / df_temp['assets'] * 100,2)
        
        if type(df_result) == int:
            df_result = df_temp[['endDate',ratio]]
        else:
            df_result = df_result.merge(df_temp[['endDate',ratio]], on=['endDate'], how='outer')
        df_result.rename(columns={ratio: stockcode}, inplace=True)
        
        


    df_result = df_result.sort(columns=['endDate'], ascending=[True])
    df_result['endDate'] = df_result['endDate'].astype(str).str[:4] + "年"
    df_result.rename(columns={'endDate': index + "相对投入资本(%)"}, inplace=True)
    df_result = df_result.set_index(index + "相对投入资本(%)")
    #df_result = df_result.dropna()
    
      
    df_result = df_result[list_ticker]
    for k,v in dict_stock.items():
        df_result.rename(columns={k: v}, inplace=True)
    
    for i in df_result:
        df_result[i] = np.nan_to_num(df_result[i])
        
    return df_result

# 得到资源占用详情
def getOccupationDetail(ticker):
    df = DataAPI.FdmtBSAllLatestGet(ticker=ticker,secID=u"",reportType=u"A",endDate=u"",beginDate=u"",year="",field=u"",pandas="1")
    df_result = 1

    index = df['secShortName'].tolist()[0]

    for i in df:
        df[i] = np.nan_to_num(df[i])
        pass

    df_temp = df
    df_temp['occupation'] = ((df_temp['AP']) + (df_temp['advanceReceipts']) + (df_temp['othPayable'])) -  ((df_temp['AR']) + (df_temp['prepayment']) + (df_temp['othReceiv'])) 
    
    df_temp['debt'] = (df_temp['STBorr'])+ (df_temp['NCLWithin1Y'])+ (df_temp['LTBorr'])+ (df_temp['bondPayable'])+ (df_temp['LTPayable'])

    df_temp['assets'] = ((df_temp['TShEquity'])+ (df_temp['STBorr'])+ (df_temp['NCLWithin1Y'])+ (df_temp['LTBorr'])+ (df_temp['bondPayable'])+ (df_temp['LTPayable']))

    df_temp = df_temp[['endDate','assets','TShEquity','debt','occupation','AP','advanceReceipts','othPayable','AR','prepayment','othReceiv',]]#'cashCEquiv']]
    
    for i in df_temp:
        if i == 'endDate':
            continue
        df_temp[i] = np.round(df_temp[i] / 100000000,2).astype(str)+ "亿"

    df_temp = df_temp.sort(columns=['endDate'],ascending=[True])
    df_temp['endDate'] = df_temp['endDate'].astype(str).str[:4]+"年"
    names = {'endDate':index,'occupation':'占用资产','assets':'投入资产','AP':'应付账款','advanceReceipts':'预收款项','othPayable':'其他应付款','AR':'应收账款','prepayment':'预付款项','othReceiv':'其他应收款','debt':'有息负债','TShEquity':'股东权益','cashCEquiv':'货币资金'}
    for k,v in names.items():
        try:
            df_temp.rename(columns={k: v}, inplace=True)
        except:
            pass

    return df_temp.set_index(index).T

# 盈利率比较
def compareByRatio(list_ticker, ratio = 'ROIC'):
    df = DataAPI.FdmtIndiRtnGet(ticker=list_ticker,secID="",endDate="",beginDate="",beginYear=u"",endYear=u"",reportType=u"A",field=u"",pandas="1")
    df_result = 1

    for i in df:
        df[i] = np.nan_to_num(df[i])
    
    dict_stock = {}
    for ticker in df.groupby('ticker'):
        df_temp = ticker[1][:11]
        stockcode =  ticker[0]
        stockname =  df_temp.secShortName.tolist()[0]
        dict_stock[stockcode] = stockname

        df_temp[ratio] = np.round(df_temp[ratio],2)
        if type(df_result) == int:
            df_result = df_temp[['endDate',ratio]]
        else:
            df_result = df_result.merge(df_temp[['endDate',ratio]], on=['endDate'], how='outer')
        df_result.rename(columns={ratio: stockcode}, inplace=True)


    df_result = df_result.sort(columns=['endDate'], ascending=[True])
    df_result['endDate'] = df_result['endDate'].astype(str).str[:4] + "年"

    if ratio == 'grossMARgin':
        ratio = "毛利率"
    if ratio == 'npMARgin':
        ratio = "净利率"

    df_result.rename(columns={'endDate': ratio + "(%)"}, inplace=True)
    df_result = df_result.set_index(ratio + "(%)")
    #df_result = df_result.dropna()
    
      
    df_result = df_result[list_ticker]
    for k,v in dict_stock.items():
        df_result.rename(columns={k: v}, inplace=True)

    for i in df_result:
        df_result[i] = np.nan_to_num(df_result[i])
        
    return df_result

# 现金流量净额
def compareByCash(list_ticker, index = '现金流量净额'):
    df = DataAPI.FdmtCFAllLatestGet(ticker=list_ticker,secID=u"",reportType=u"A",endDate=u"",beginDate=u"",year="",field=u"",pandas="1")
    df_result = 1
    
    dict_type = {'现金流量净额':'NChangeInCash',
                '经营现金流':'NCFOperateA',
                '投资现金流':'NCFFrInvestA',
                '筹资现金流':'NCFFrFinanA'}
    
    
    ratio = dict_type[index]
    
    # NCFOperateA	float	经营活动产生的现金流量净额
    # NCFFrInvestA	float	投资活动产生的现金流量净额
    # NCFFrFinanA	float	筹资活动产生的现金流量净额

    for i in df:
        df[i] = np.nan_to_num(df[i])
    
    dict_stock = {}
    for ticker in df.groupby('ticker'):
        df_temp = ticker[1][:11]
        stockcode =  ticker[0]
        stockname =  df_temp.secShortName.tolist()[0]
        dict_stock[stockcode] = stockname

        df_temp[ratio] = np.round(df_temp[ratio] / 1000000.0,2)
        if type(df_result) == int:
            df_result = df_temp[['endDate',ratio]]
        else:
            df_result = df_result.merge(df_temp[['endDate',ratio]], on=['endDate'], how='outer')
        df_result.rename(columns={ratio: stockcode}, inplace=True)


    df_result = df_result.sort(columns=['endDate'], ascending=[True])
    df_result['endDate'] = df_result['endDate'].astype(str).str[:4] + "年"


    df_result.rename(columns={'endDate': index + "（百万）"}, inplace=True)
    df_result = df_result.set_index(index + "（百万）")
    #df_result = df_result.dropna()
    
    df_result = df_result[list_ticker]
    for k,v in dict_stock.items():
        df_result.rename(columns={k: v}, inplace=True)

    for i in df_result:
        df_result[i] = np.nan_to_num(df_result[i])
        
    return df_result

# 三费占比
def compareByThreeFee(list_ticker, ratio = "三费占比"):
    df = DataAPI.FdmtISAllLatestGet(ticker=list_ticker,secID=u"",reportType=u"A",year="",endDate=u"",beginDate=u"",field=u"",pandas="1")
    df_result = 1
    
    for i in df:
        df[i] = np.nan_to_num(df[i])
    
    dict_stock = {}
    for ticker in df.groupby('ticker'):
        df_temp = ticker[1][:10]
        stockcode =  ticker[0]
        stockname =  df_temp.secShortName.tolist()[0]
        dict_stock[stockcode] = stockname
        
        if ratio == '三费占比':
            df_temp[ratio]  = (df_temp['sellExp'] + df_temp['adminExp'] + df_temp['finanExp']) / df_temp['revenue'] * 100
        if ratio =='销售费用占比':
            df_temp[ratio]  = df_temp['sellExp'] / df_temp['revenue'] * 100
        if ratio =='管理费用占比':
            df_temp[ratio]  = df_temp['adminExp'] / df_temp['revenue'] * 100
        if ratio =='财务费用占比':
            df_temp[ratio]  = df_temp['finanExp'] / df_temp['revenue'] * 100
            
        df_temp[ratio] = np.round(df_temp[ratio],2)
        if type(df_result) == int:
            df_result = df_temp[['endDate',ratio]]
        else:
            df_result = df_result.merge(df_temp[['endDate',ratio]], on=['endDate'], how='outer')
        df_result.rename(columns={ratio: stockcode}, inplace=True)


    df_result = df_result.sort(columns=['endDate'], ascending=[True])
    df_result['endDate'] = df_result['endDate'].astype(str).str[:4] + "年"
    df_result.rename(columns={'endDate': ratio + "(%)"}, inplace=True)
    df_result = df_result.set_index(ratio + "(%)")
    #df_result = df_result.dropna()
    
      
    df_result = df_result[list_ticker]
    for k,v in dict_stock.items():
        df_result.rename(columns={k: v}, inplace=True)

    for i in df_result:
        df_result[i] = np.nan_to_num(df_result[i])
        
    return df_result

# 周转情况
def compareByTurnover(list_ticker, index = "存货周转次数"):
    df_BS = DataAPI.FdmtBSAllLatestGet(ticker=list_ticker,secID=u"",reportType=u"A",endDate=u"",beginDate=u"",year="",field=u"ticker,secShortName,endDate,inventories,AR,TCA,TAssets",pandas="1")
    df_IS = DataAPI.FdmtISAllLatestGet(ticker=list_ticker,secID=u"",reportType=u"A",year="",endDate=u"",beginDate=u"",field=u"ticker,secShortName,endDate,COGS,revenue",pandas="1")
    df = df_BS.merge(df_IS, on=['ticker','secShortName','endDate'], how='outer')
    
    for i in df:
        df[i] = np.nan_to_num(df[i])
    
    df_result = 1
    
    dict_stock = {}
    for ticker in df.groupby('ticker'):
        df_temp = ticker[1][:11]
        stockcode =  ticker[0]
        stockname =  df_temp.secShortName.tolist()[0]
        dict_stock[stockcode] = stockname
        
        if index == "存货周转次数" or index == "存货周转天数":
            # 存货周转次数=营业成本（COGS）/平均存货     【平均存货=(上一年度期末存货+期末存货（inventories）)/2】
            df_temp['preInventories'] = df_temp['inventories']
            df_temp['preInventories'][df_temp.index] = df_temp['inventories'][df_temp.index+1]
            df_temp[index] = df_temp['COGS']*2/(df_temp['preInventories'] + df_temp['inventories'])
            if index == "存货周转天数":
                df_temp[index] = 365 / df_temp[index]
        
        if index == "应收账款周转次数" or index == "应收账款周转天数":
            # 应收账款周转次数=营业收入(revenue)/平均应收账款 ，平均应收账款=（期初应收账款+期末应收账款(AR)）/2，反映应收账款转换为现金的次数
            df_temp['preAR'] = df_temp['AR']
            df_temp['preAR'][df_temp.index] = df_temp['AR'][df_temp.index+1]
            df_temp[index] = df_temp['revenue']*2/(df_temp['preAR'] + df_temp['AR'])
            if index == "应收账款周转天数":
                df_temp[index] = 365 / df_temp[index]
        
        if index == "总资产周转次数":
            # 总资产周转次数=营业收入（revenue）/总资产平均余额     【总资产平均余额=（期初总资产+期末总资产（TAssets））/2】
            df_temp['preTAssets'] = df_temp['TAssets']
            df_temp['preTAssets'][df_temp.index] = df_temp['TAssets'][df_temp.index+1]
            df_temp[index] = df_temp['revenue']*2/(df_temp['preTAssets'] + df_temp['TAssets'])
        
        if index == "流动资产周转次数":
            # 流动资产周转次数=营业收入（revenue）/流动资产平均余额     【总资产平均余额=（期初流动资产+期末流动资产（TCA））/2】
            df_temp['preTCA'] = df_temp['TCA']
            df_temp['preTCA'][df_temp.index] = df_temp['TCA'][df_temp.index+1]
            df_temp[index] = df_temp['revenue']*2/(df_temp['preTCA'] + df_temp['TCA'])
        
        
        
        df_temp[index] = np.round(df_temp[index],2)
        
        if type(df_result) == int:
            df_result = df_temp[['endDate',index]]
        else:
            df_result = df_result.merge(df_temp[['endDate',index]], on=['endDate'], how='outer')
        df_result.rename(columns={index: stockcode}, inplace=True)
        
        


    df_result = df_result.sort(columns=['endDate'], ascending=[True])
    df_result['endDate'] = df_result['endDate'].astype(str).str[:4] + "年"
    df_result.rename(columns={'endDate': index}, inplace=True)
    df_result = df_result.set_index(index)
    #df_result = df_result.dropna()
    
      
    df_result = df_result[list_ticker]
    for k,v in dict_stock.items():
        df_result.rename(columns={k: v}, inplace=True)

    for i in df_result:
        df_result[i] = np.nan_to_num(df_result[i])
        
    return df_result

# 偿债能力
def compareByDebtPayingAbility(list_ticker, index):
    df_BS = DataAPI.FdmtBSAllLatestGet(ticker=list_ticker,secID=u"",reportType=u"A",endDate=u"",beginDate=u"",year="",field=u"ticker,secShortName,endDate,inventories,TCA,TCL,TLiab,TAssets,AP,advanceReceipts,othPayable,AR,prepayment,othReceiv",pandas="1")
    df_IS = DataAPI.FdmtISAllLatestGet(ticker=list_ticker,secID=u"",reportType=u"A",year="",endDate=u"",beginDate=u"",field=u"ticker,secShortName,endDate,NIncome,finanExp,incomeTax",pandas="1")
    
    df = df_BS.merge(df_IS, on=['ticker','secShortName','endDate'], how='outer')
    
    for i in df:
        df[i] = np.nan_to_num(df[i])
    
    
    df_result = 1
    
    dict_stock = {}
    for ticker in df.groupby('ticker'):
        df_temp = ticker[1][:10]
        stockcode =  ticker[0]
        stockname =  df_temp.secShortName.tolist()[0]
        dict_stock[stockcode] = stockname
        
        if index == '流动比率':
            # 流动比率 = 流动资产（TCA） ÷ 流动负债（TCL）
            df_temp[index] = df_temp['TCA'] / df_temp['TCL']
            
        if index == '速动比率':
            # 速动比率 = （流动资产（TCA） - 存货（inventories））÷ 流动负债（TCL）
            df_temp[index] = (df_temp['TCA'] - df_temp['inventories'] )/ df_temp['TCL']
        
        if index == '资产负债率':
            # 资产负债率 = 总负债（TLiab） ÷ 总资产（TAssets）
            df_temp[index] = df_temp['TLiab'] / df_temp['TAssets']
            
        if index == '扣除良性负债的资产负债率':
            df_temp['occupation'] = ((df_temp['AP']) + (df_temp['advanceReceipts']) + (df_temp['othPayable'])) -  ((df_temp['AR']) + (df_temp['prepayment']) + (df_temp['othReceiv'])) 
            df_temp[index] = df_temp['occupation'] / df_temp['TAssets']
        
        if index == '利息保障倍数':
            # 利息保障倍数 = （净利润（NIncome） + 利息（finanExp） + 税（incomeTax）） ÷ 利息（intExp）
            df_temp[index] = (df_temp['NIncome'] + df_temp['finanExp'] + df_temp['incomeTax'] )/ df_temp['finanExp']
            
        df_temp[index] = np.round(df_temp[index],2)
                                        
        if type(df_result) == int:
            df_result = df_temp[['endDate',index]]
        else:
            df_result = df_result.merge(df_temp[['endDate',index]], on=['endDate'], how='outer')
        df_result.rename(columns={index: stockcode}, inplace=True)
                

    df_result = df_result.sort(columns=['endDate'], ascending=[True])
    df_result['endDate'] = df_result['endDate'].astype(str).str[:4] + "年"
    df_result.rename(columns={'endDate': index}, inplace=True)
    df_result = df_result.set_index(index)
    #df_result = df_result.dropna()
    for i in df_result:
        df_result[i] = np.nan_to_num(df_result[i])
      
    df_result = df_result[list_ticker]
    for k,v in dict_stock.items():
        df_result.rename(columns={k: v}, inplace=True)

    for i in df_result:
        df_result[i] = np.nan_to_num(df_result[i])
        
    return df_result


def getRanks(df_result, ascending = False):
    
    avg_total = df_result.mean()
    avg_five = df_result.tail(5).mean()
    
    df_result_T = df_result.T
    
    df_result_T['rank'] = 1
    df_result_T['rank'] =pd.rolling_sum(df_result_T['rank'], window = len(df_result_T),min_periods = 1)
    df_result_T['rank'] = df_result_T['rank'].astype(int)

    df_result_T['平均'] = avg_total
    df_result_T['平均'] = np.round(df_result_T['平均'],2)
    df_result_T = df_result_T.sort(columns=['平均'],ascending=[ascending])
    df_result_T['排名'] = 1
    df_result_T['排名'] =pd.rolling_sum(df_result_T['排名'], window = len(df_result_T),min_periods = 1)
    df_result_T['排名'] = df_result_T['排名'].astype(int)
    
    df_result_T['近5年平均'] = avg_five
    df_result_T['近5年平均'] = np.round(df_result_T['近5年平均'],2)
    df_result_T = df_result_T.sort(columns=['近5年平均'],ascending=[ascending])
    df_result_T['近5年排名'] = 1
    df_result_T['近5年排名'] =pd.rolling_sum(df_result_T['近5年排名'], window = len(df_result_T),min_periods = 1)
    df_result_T['近5年排名'] = df_result_T['近5年排名'].astype(int)
  
    df_result_T = df_result_T.sort(columns=['rank'],ascending=[True])
    df_result_T.drop(['rank'], axis=1, inplace=True)

    return df_result_T

# 三费占比（先详细）
def getThreeFee(list_ticker):
    df = DataAPI.FdmtISAllLatestGet(ticker=list_ticker,secID=u"",reportType=u"A",year="",endDate=u"",beginDate=u"",field=u"",pandas="1")
    
    df['销售费用占比']  = np.round(df['sellExp'] / df['revenue'] * 100,2).astype(str) + "%"
    df['管理费用占比']  = np.round(df['adminExp'] / df['revenue'] * 100,2).astype(str) + "%"
    df['财务费用占比']  = np.round(df['finanExp'] / df['revenue'] * 100,2).astype(str) + "%"
    df['三费占比']  = np.round((df['sellExp'] + df['adminExp'] + df['finanExp']) / df['revenue'] * 100,2).astype(str) + "%"
    df['sellExp'] = np.round(df['sellExp'] / 1000000,2).astype(str) + "(百万)"
    df['adminExp'] = np.round(df['adminExp'] / 1000000,2).astype(str) + "(百万)"
    df['finanExp'] = np.round(df['finanExp'] / 1000000,2).astype(str) + "(百万)"
    df['revenue'] = np.round(df['revenue'] / 1000000,2).astype(str) + "(百万)"
    
    stockname =  df.secShortName.tolist()[0]
    
    df = df.sort(columns=['endDate'], ascending=[True])
    df['endDate'] = df['endDate'].astype(str).str[:4] + "年"
    df.rename(columns={'endDate': stockname}, inplace=True)
    df = df.set_index(stockname)
    
    df.rename(columns={'sellExp':'销售费用','adminExp':'管理费用','finanExp':'财务费用','revenue':'营业收入'}, inplace=True)
    
    df = df[['营业收入','销售费用','管理费用','财务费用','销售费用占比','管理费用占比','财务费用占比','三费占比']]
    df = df.dropna()
    

    return df.T

# 投入产出比
def compareByInvest(list_ticker, index = 'investOut'):
    df = DataAPI.FdmtCFAllLatestGet(ticker=list_ticker,secID=u"",reportType=u"A",endDate=u"",beginDate=u"",year="",field=u"",pandas="1")
    
    df_IS = DataAPI.FdmtISAllLatestGet(ticker=list_ticker,secID=u"",reportType=u"A",year="",endDate=u"",beginDate=u"",field=u"ticker,secShortName,endDate,NIncome,finanExp,incomeTax",pandas="1")
    
    df = df.merge(df_IS, on=['ticker','secShortName','endDate'], how='outer')
    for i in df:
        df[i] = np.nan_to_num(df[i])
    
    if index == 'investOut':
        columns = ['证券名称','十年资本支出之和','十年净现金流之和','产出投入比']
    if index == 'investIncrease':
        columns = ['证券名称','十年资本支出之和','十年EBIT增长','增长投入比']

    
    df_result = pd.DataFrame(np.zeros(len(list_ticker)*len(columns)).reshape(len(list_ticker),len(columns)),index = list_ticker,columns = columns)

    for ticker in df.groupby('ticker'):
        df_temp = ticker[1][:10]
        stockcode =  ticker[0]
        stockname =  df_temp.secShortName.tolist()[0]
        
        # NCFOperateA	float	经营活动产生的现金流量净额
        # NCFFrInvestA	float	投资活动产生的现金流量净额
        # cashflow 净现金流
        
        df_temp['cashflow'] = df_temp['NCFOperateA'] - df_temp['NCFFrInvestA']
        invest = - np.round(df_temp['NCFFrInvestA'].sum() / 100000000,2)
        out =  np.round(df_temp['cashflow'].sum() /100000000,2)
        df_temp['ebit'] = df_temp['NIncome'] + df_temp['finanExp'] + df_temp['incomeTax']
        
        list_ebit = df_temp['ebit'].tolist()
        increase = round((list_ebit[0]-list_ebit[-1])/100000000,2)
        
        investOut = - np.round(df_temp['cashflow'].sum() / df_temp['NCFFrInvestA'].sum(),2)
        
        investIncrease = - np.round((list_ebit[0]-list_ebit[-1]) / df_temp['NCFFrInvestA'].sum(),2)
 
        if index == 'investOut':
            df_result.ix[stockcode] = [stockname,"%s（亿）" % invest,"%s（亿）" % out,investOut]
        if index == 'investIncrease':
            df_result.ix[stockcode] = [stockname,"%s（亿）" % invest,"%s（亿）" % increase,investIncrease]
            
            
    return df_result

# 得到分红情况
def getCashDiv(ticker):
    df_fh = DataAPI.EquDivGet(exDivDate="",eventProcessCD=u"",secID=u"",ticker=ticker,beginDate=u"",endDate=u"",field=u"",pandas="1")
    df_fh = df_fh[df_fh.endDate>"2007-12-31"][['endDate','perShareDivRatio','perCashDiv','perCashDivAfTax','payCashDate']]

    df_fh['endDate'] = df_fh['endDate'].astype(str).str[:4] + "年"
    df_fh['perShareDivRatio'][np.isnan(df_fh['perShareDivRatio'])] = '未送股'
    df_fh['perCashDiv'][np.isnan(df_fh['perCashDiv'])] = '未分红'
    df_fh['perCashDivAfTax'][np.isnan(df_fh['perCashDivAfTax'])] = '未分红'
    df_fh['payCashDate'] = df_fh['payCashDate'].astype(str)
    
    df_fh['payCashDate'][df_fh['payCashDate'] == "nan"] = '未分红'
    fh = df_fh.perCashDiv.tolist()[-1]
    
    try:
        price = DataAPI.MktEqudGet(tradeDate=u"",secID=u"",ticker=ticker,beginDate=u"",endDate=u"",isOpen="",field=u"closePrice",pandas="1").closePrice.tolist()[-1]
        pct = fh / price *100
        print "最新年度税前分红占收盘价比例 %.2f%%" % pct
    except:
        pass

    names = {'endDate':'分红年度','perShareDivRatio':'每股送股比例','perCashDiv':'每股派现(税前)','perCashDivAfTax':'每股派现(税后)','payCashDate':'派现日',}
    for k,v in names.items():
        df_fh.rename(columns={k: v}, inplace=True)

    df_fh = df_fh.set_index('分红年度')

    tradeDateStr = pd.datetime.today().strftime("%Y%m%d")
    

    return df_fh.T




from CAL.PyCAL import *    # CAL.PyCAL中包含font
from matplotlib import pylab
import matplotlib.pyplot as plt
def drawByDF(df, title = "ROIC"):
    
    fig, ax = plt.subplots(1, figsize=(22,8))
    fig.set_facecolor('white')
    colormap = plt.cm.Paired
    ax = plt.gca()
    ax.set_color_cycle([colormap(i) for i in np.linspace(0.1, 1.0, len(df.T))])
    
    num = 1
    for i in df:
        label = i.decode('utf-8')
        if num == 1:
            ax.plot(df[i].tolist(),label = label,lw=3)
        else:
            ax.plot(df[i].tolist(),label = label,lw=1)
        num += 1
    
    
    patches, labels = ax.get_legend_handles_labels()
    font.set_size(18)
    legend = ax.legend(patches, labels, loc=2, prop = font )  # 使用font
    frame = legend.get_frame()  
    frame.set_alpha(1)  
    frame.set_facecolor('white') # 设置图例legend背景透明  
    
 
    ax.set_title(title, fontproperties = font, fontsize = 20)  # 使用font
    ax.set_xlabel(u'时间', fontproperties = font, fontsize = 15)  # 使用font 
    
    xlabellist = [u'%s年' % x[:4] for x in df.index.tolist()]

    ax.set_xticklabels(xlabellist, fontproperties = font, fontsize=15)  
    
    if title in ["ROE"]:
        ax.set_ylim(-20, 50)
    
    ax.axhline(y=0,linewidth=0.5, color='black',ls='--')
    
```
