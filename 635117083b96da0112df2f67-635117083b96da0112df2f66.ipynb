```python
import pandas as pd

# 获取上证综指的历史日行情
df1 = DataAPI.MktIdxdGet(indexID=u"",ticker=['000001'], tradeDate=u"",beginDate=u"",endDate=u"",exchangeCD=u"XSHE,XSHG",field=["ticker", "secShortName", "tradeDate", "closeIndex"],pandas="1")

# 获取深成综指的历史日行情
df2= DataAPI.MktIdxdGet(indexID=u"",ticker=['399106'], tradeDate=u"",beginDate=u"",endDate=u"",exchangeCD=u"XSHE,XSHG",field=["ticker", "secShortName", "tradeDate", "closeIndex"],pandas="1")

df_meeting = pd.read_csv('./meeting.csv')
df_trade = pd.read_csv('./trade_day.csv', index_col = 0)

# 转换函数，将开会日期若是非交易日，则开始日前移到交易日，结束日后移到交易日
def convert_day(begintrade, endtrade):
    df_begin = df_trade.loc[df_trade['calendarDate'] == begintrade, :]
    df_end = df_trade.loc[df_trade['calendarDate'] == endtrade, :]
    while df_begin.empty:
        begintrade = begintrade - 1
        df_begin = df_trade.loc[df_trade['calendarDate'] == begintrade, :]
    act_begin = begintrade

    while df_end.empty:
        endtrade = endtrade + 1
        df_end = df_trade.loc[df_trade['calendarDate'] == endtrade, :]
    act_end = endtrade
    td = [act_begin, act_end]
    return act_begin, act_end

# 获取 大会前和大会后 N 日的交易日期
def before_after(N, act_begin, act_end):
    df_trade['before'] = df_trade['calendarDate'].shift(N)
    df_trade['after'] = df_trade['calendarDate'].shift(-N)
    df_trade['before'].fillna(0,  inplace = True)
    df_trade['before'] = df_trade['before'].astype(int)
    df_trade['after'].fillna(0,  inplace = True)
    df_trade['after'] = df_trade['after'].astype(int)
    df_before = df_trade.loc[df_trade['calendarDate'] == act_begin, :]
    df_after = df_trade.loc[df_trade['calendarDate'] == act_end, :]
    df_before = df_before.reset_index(drop=True)
    df_after = df_after.reset_index(drop=True)
    before_day = df_before['before'][0]
    after_day = df_after['after'][0]
    return before_day, after_day

df_meeting['actbegin'] = 0
df_meeting['actend'] = 0

df_meeting['ShangHai_before'] = 0
df_meeting['ShangHai_begin'] = 0
df_meeting['ShangHai_end'] = 0
df_meeting['ShangHai_after'] = 0

df_meeting['ShenZhen_before'] = 0
df_meeting['ShenZhen_begin'] = 0
df_meeting['ShenZhen_end'] = 0
df_meeting['ShenZhen_after'] = 0
df_meeting['before_day'] = 0
df_meeting['after_day'] = 0

for i in range(7):
    # 开会期间，开始日期begintrade， 结束日期endtrade；
    begintrade = df_meeting['begin'][i]
    endtrade = df_meeting['end'][i]
    # 若是周末，则 向前 / 后移动一个交易日
    act_begin, act_end = convert_day(begintrade, endtrade)
    df_meeting['actbegin'][i] = act_begin
    df_meeting['actend'][i] = act_end
    df_SH_begin = DataAPI.MktIdxdGet(indexID=u"",ticker=['000001'], tradeDate = str(act_begin), beginDate=u"",endDate=u"",exchangeCD=u"XSHE,XSHG",field=["ticker", "secShortName", "tradeDate", "closeIndex"],pandas="1")
    
    df_meeting['ShangHai_begin'][i] = df_SH_begin['closeIndex'][0]
    df_SH_end =  DataAPI.MktIdxdGet(indexID=u"",ticker=['000001'], tradeDate= str(act_end), beginDate=u"",endDate=u"",exchangeCD=u"XSHE,XSHG",field=["ticker", "secShortName", "tradeDate", "closeIndex"],pandas="1")
    if not df_SH_end.empty:
        df_meeting['ShangHai_end'][i] = df_SH_end['closeIndex'][0]
    
    df_SZ_begin = DataAPI.MktIdxdGet(indexID=u"",ticker=['399106'], tradeDate=str(act_begin), beginDate=u"",endDate=u"",exchangeCD=u"XSHE,XSHG",field=["ticker", "secShortName", "tradeDate", "closeIndex"],pandas="1")
    df_meeting['ShenZhen_begin'][i] = df_SZ_begin['closeIndex'][0]
    df_SZ_end = DataAPI.MktIdxdGet(indexID=u"",ticker=['399106'], tradeDate=str(act_end), beginDate=u"",endDate=u"",exchangeCD=u"XSHE,XSHG",field=["ticker", "secShortName", "tradeDate", "closeIndex"],pandas="1")
    if not df_SZ_end.empty:
        df_meeting['ShenZhen_end'][i] = df_SZ_end['closeIndex'][0]
    # 开会前 N 个交易日 before_day 和 开会后 N个交易日 after_day
    N = 5
    before_day, after_day = before_after(N, act_begin, act_end)

    df_meeting['before_day'][i] = before_day
    df_meeting['after_day'][i] = after_day

    df_SH_before = DataAPI.MktIdxdGet(indexID=u"",ticker=['000001'], tradeDate = str(before_day), beginDate=u"",endDate=u"",exchangeCD=u"XSHE,XSHG",field=["ticker", "secShortName", "tradeDate", "closeIndex"],pandas="1")
    if not df_SH_before.empty:  
        df_meeting['ShangHai_before'][i] = df_SH_before['closeIndex'][0]
    
    df_SH_after = DataAPI.MktIdxdGet(indexID=u"",ticker=['000001'], tradeDate = str(after_day), beginDate=u"",endDate=u"",exchangeCD=u"XSHE,XSHG",field=["ticker", "secShortName", "tradeDate", "closeIndex"],pandas="1")
    if not df_SH_after.empty: 
        df_meeting['ShangHai_after'][i] = df_SH_after['closeIndex'][0]

    df_SZ_before = DataAPI.MktIdxdGet(indexID=u"",ticker=['399106'], tradeDate = str(before_day), beginDate=u"",endDate=u"",exchangeCD=u"XSHE,XSHG",field=["ticker", "secShortName", "tradeDate", "closeIndex"],pandas="1")
    if not df_SZ_before.empty:  
        df_meeting['ShenZhen_before'][i] = df_SZ_before['closeIndex'][0]
    
    df_SZ_after = DataAPI.MktIdxdGet(indexID=u"",ticker=['399106'], tradeDate = str(after_day), beginDate=u"",endDate=u"",exchangeCD=u"XSHE,XSHG",field=["ticker", "secShortName", "tradeDate", "closeIndex"],pandas="1")
    if not df_SZ_after.empty: 
        df_meeting['ShenZhen_after'][i] = df_SZ_after['closeIndex'][0]

df_meeting['SH_return'] = df_meeting['ShangHai_end']/ df_meeting['ShangHai_begin'] -1
df_meeting['SH_return'] = df_meeting['SH_return'].apply(lambda x: format(x, '.2%'))
df_meeting['SZ_return'] = df_meeting['ShenZhen_end']/ df_meeting['ShenZhen_begin'] -1
df_meeting['SZ_return'] = df_meeting['SZ_return'].apply(lambda x: format(x, '.2%'))

df_meeting['SH_before_return'] = df_meeting['ShangHai_begin']/ df_meeting['ShangHai_before'] -1
df_meeting['SH_before_return'] = df_meeting['SH_before_return'].apply(lambda x: format(x, '.2%'))
df_meeting['SH_after_return'] = df_meeting['ShangHai_after']/ df_meeting['ShangHai_end'] -1
df_meeting['SH_after_return'] = df_meeting['SH_after_return'].apply(lambda x: format(x, '.2%'))

df_meeting['SZ_before_return'] = df_meeting['ShenZhen_begin']/ df_meeting['ShenZhen_before'] -1
df_meeting['SZ_before_return'] = df_meeting['SZ_before_return'].apply(lambda x: format(x, '.2%'))
df_meeting['SZ_after_return'] = df_meeting['ShenZhen_after']/ df_meeting['ShenZhen_end'] -1
df_meeting['SZ_after_return'] = df_meeting['SZ_after_return'].apply(lambda x: format(x, '.2%'))
df_meeting[['name', 'before_day', 'ShangHai_before', 'ShenZhen_before', 'begin', 'actbegin', 'ShangHai_begin', 'ShenZhen_begin', 'end', 'actend', 'ShangHai_end', 'ShenZhen_end', 'after_day', 'ShangHai_after', 'ShenZhen_after']]
df_return = df_meeting.loc[:, ['name','before_day' , 'actbegin', 'actend','after_day','SH_before_return', 'SZ_before_return', 'SH_return', 'SZ_return', 'SH_after_return', 'SZ_after_return']]
df_return
```
