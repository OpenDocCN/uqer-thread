# 市场情绪分析--分级基金折溢价

分级基金是目前市场上门槛最低，参与人数最多的衍生品，在2014年开启的**那波**牛市，风光无限。

市场玩家众多，有长期持有A类的，轮动的，折溢价套利的，买B搏一把的，上交所分级T+0的，还有交学费的，等等等。

本篇对分级基金的折溢价进行分析，共同探讨对市场情绪的反应情况。

```python
import pandas as pd
import numpy as np
from CAL.PyCAL import font
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('white')

pd.options.display.max_columns = 1000
pd.options.display.max_rows = 1000
DataAPI.settings.cache_enabled = True

start = "20150101"
```

```python
# 获取沪深两市的分级基金信息
fund_lvg_df = DataAPI.FundLeverageInfoGet(secID=u"",ticker=u"",exchangeCDLeverage=u"XSHG,XSHE",field=u"",pandas="1")
print '总共有%s只分级基金。' %(len(fund_lvg_df))
fund_lvg_df.head()
```

```python
# 陆家嘴事件以来，比较火的券商B
DataAPI.MktFunddGet(secID='150201.XSHE', beginDate='20160521')
```

```python
# 不同分级基金具有不同利率等条款，有一些分级甚至是无下折的
# 无下折的分级b包含无期限的看涨期权，因此通常具有较高溢价，本贴计算折溢价均是直接平均，由于其溢价影响较大，分析时剔除之。
fund_lvg_df[fund_lvg_df['splitNote'].str.contains('无下折').fillna(False)].secIDLeverage
```

```python
# 获取hs300指数行情数据
hs300idx = DataAPI.MktIdxdGet(ticker='000300', beginDate=start)
```

```python
# 找出非“无下折”的a，b基金的列表
lvg_a_secids = fund_lvg_df[(fund_lvg_df['shareType']=='A')&(~ fund_lvg_df['splitNote'].str.contains('无下折').fillna(False))].secIDLeverage.values
lvg_b_secids = fund_lvg_df[(fund_lvg_df['shareType']=='B')&(~ fund_lvg_df['splitNote'].str.contains('无下折').fillna(False))].secIDLeverage.values
lvg_a_df = DataAPI.MktFunddGet(tradeDate=u"",secID=lvg_a_secids,ticker=u"",beginDate=start,endDate=u"",field=u"",pandas="1")

# 成交额大于1000w，计入统计
lvg_a_df = lvg_a_df[lvg_a_df['turnoverValue']>1e7]
```

```python
# 计算每天的平均折价，并画图
discount_df = pd.DataFrame(lvg_a_df.groupby('tradeDate')['discountRatio'].mean())
discount_df['closeIndex'] = hs300idx['closeIndex'].values
discount_df.index = pd.to_datetime(discount_df.index)

fig = plt.figure(figsize=(12,6))

# 绿色曲线表示hs300走势
ax1 = fig.add_subplot(111)
ax1.plot(discount_df.index, discount_df.closeIndex, 'g', label="hs300")
ax1.legend(loc=2)
ax1.set_ylabel('hs300')

# 红色曲线表示分级a的平均折价走势
ax2 = ax1.twinx()
ax2.plot(discount_df.index, discount_df.discountRatio, 'r', label = "a discount ratio")
ax2.legend(loc=1)
ax2.set_ylabel('discount ratio')
ax2.set_xlabel('date')
plt.show()
```

```python
data=[]
for i in xrange(-20,10):
    dr = lvg_a_df.groupby('tradeDate')['discountRatio'].mean().shift(i).values
    df = pd.DataFrame({'index': hs300idx.closeIndex/hs300idx.closeIndex[0], 'dr':dr})
    data.append ((i, df.corr()['index'][0]))
corr_df = pd.DataFrame(data,columns=['shift_index', 'corr'])
```

```python
corr_df.sort(['corr'], ascending=False).head()
```

```python
corr_df.plot(x='shift_index')
```

**根据图表可以看出，指数往后偏移11个交易日与折价相关性最高。也就是说，分级a的折价滞后指数10个交易日左右。根据相关系数和指数走势图，可以看出广大股民具有强烈的抄底欲望，虽然好几次抄在腰上。**

```python
# 大于500w成交额 分级b行情数据
lvg_b_df = DataAPI.MktFunddGet(tradeDate=u"",secID=lvg_b_secids,ticker=u"",beginDate=start,endDate=u"",field=u"",pandas="1")
lvg_b_df = lvg_b_df[lvg_b_df['turnoverValue']>5e6]
```

```python
# 平均折价走势图
lvg_b_df.groupby('tradeDate')['discountRatio'].mean().plot(figsize=(16,6))
```

```python
# 溢价最高（折价最低）的20只
b_df = pd.DataFrame(lvg_b_df.groupby('tradeDate')['discountRatio'].mean().sort(inplace=False).head(20))
b_df['tradeDate']=b_df.index
hs300_df = hs300idx[hs300idx.tradeDate.isin(b_df.index.values) ][['tradeDate','CHGPct']]
# 加上当天hs300的涨跌幅
pd.merge(hs300_df, b_df, on=['tradeDate'])
```

分级b溢价最高的时间段发生在股灾1.0，2.0期间，很大程度上是因为千股跌停，分级b丧失流动性导致。悲壮程度远比昨天的398手IF1606空单砸到跌停。

目前市场风险偏好极低，昨天（20160531）分级b竟然是折价（按照上面的加权平均算法），中短期看来，市场情绪仍然悲观。
