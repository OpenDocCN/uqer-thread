## 2. 复星概念股走势跟踪

```python
# ！！！ 复星控制的五家公司股票
fosun_data = pd.DataFrame()
hs300 = DataAPI.MktIdxdGet(ticker='000300', beginDate='20151208', endDate='20151214', field='closeIndex', pandas='1')
fosun_data[u'沪深300'] = hs300.closeIndex/hs300.closeIndex.values[-3]-1
for sec in fosun.index[0:5]:
    fields = 'secID,tradeDate,closePrice'
    tmp = DataAPI.MktEqudGet(secID=sec, beginDate='20151208', endDate='20151214', field=fields, pandas='1')
    fosun_data['tradeDate'] = tmp.tradeDate
    print 
    fosun_data[fosun.secShortName[sec]] = tmp.closePrice/tmp.closePrice.values[-3]-1.0
fosun_data = fosun_data.set_index('tradeDate')
fosun_data.index = map(Date.toDateTime, map(Date.parseISO, fosun_data.index))
pylab.figure(figsize=(10,6))
for column in fosun_data.columns:
    pylab.plot(fosun_data.index, fosun_data[column],'-o')

pylab.legend(fosun_data.columns, prop = font)
pylab.title(u'复星控股的股票最近走势', fontproperties = font)
pylab.show()

# ！！！ 复星作为第二大股东的公司股票
fosun_data = pd.DataFrame()
hs300 = DataAPI.MktIdxdGet(ticker='000300', beginDate='20151208', endDate='20151214', field='closeIndex', pandas='1')
fosun_data[u'沪深300'] = hs300.closeIndex/hs300.closeIndex.values[-3]-1
for sec in fosun.index[5:12]:
    fields = 'secID,tradeDate,closePrice'
    tmp = DataAPI.MktEqudGet(secID=sec, beginDate='20151208', endDate='20151214', field=fields, pandas='1')
    fosun_data['tradeDate'] = tmp.tradeDate
    print 
    fosun_data[fosun.secShortName[sec]] = tmp.closePrice/tmp.closePrice.values[-3]-1.0
fosun_data = fosun_data.set_index('tradeDate')
fosun_data.index = map(Date.toDateTime, map(Date.parseISO, fosun_data.index))
pylab.figure(figsize=(10,6))
for column in fosun_data.columns:
    pylab.plot(fosun_data.index, fosun_data[column],'-o')

pylab.legend(fosun_data.columns, prop = font)
pylab.title(u'复星作为二股东的股票最近走势', fontproperties = font)
pylab.show()
```

图中，我们给出了沪深300指数和复星相关股票近期走势图，图中纵轴为个股股价相对于12月10日（复星停牌前一天收盘价）的涨跌幅
- 五家复星控股公司上周五停牌，除海南矿业外今日均大跌，受事件影响比较大
- 复星作为二股东的7家，周五随指数下跌，今日除汉森制药外均随指数有所回升，受事件影响比较小
- 复星控股的5家公司，今日股票市场交易量均有所放大
