# 指数 vs ETF，它们的变化有区别吗？

# 拿沪深300指数当做benchmark好不好的一个问题

一句话来说，就是通常的沪深300指数的变化会比实际情况略微差一点，因为沪深300指数对于指数中的成分股分红，不予修正。而如果买了ETF，选择红利再投资的话，则就相对于指数略微表现好了一点。然后ETF有少许费用，但是应该敌不过分红吧，毕竟这是沪深300……

实际情况如最后的图片。似乎这两年的股息率有点高。毛估估的话，在3%左右。（不计ETF因为其他影响导致跟踪误差的话……）

```python
hs300_idx = DataAPI.MktIdxdGet(tradeDate=u"",indexID=u"000300.ZICN",ticker=u"",beginDate=u"",endDate=u"",exchangeCD=u"XSHE,XSHG",field=u"",pandas="1")
hs300_idx = hs300_idx[['tradeDate','closeIndex']]
hs300_idx['tradeDate'] = pd.to_datetime(hs300_idx['tradeDate'])
hs300_idx.set_index('tradeDate',inplace=True)
```

```python
hs300_etf = DataAPI.FundNavGet(dataDate=u"",secID=u"",ticker=u"510300",beginDate=u"",endDate=u"",field=u"",pandas="1")
hs300_etf = hs300_etf[['endDate','ADJUST_NAV']]
hs300_etf['endDate'] = pd.to_datetime(hs300_etf['endDate'])
hs300_etf.set_index('endDate',inplace=True)
```

```python
mg = pd.concat([hs300_idx,hs300_etf],axis=1)
mg = mg.dropna()
mg = mg.rename(columns={'closeIndex':'idx','ADJUST_NAV':'etf'})
(1+mg.pct_change()).cumprod().plot(figsize=(12,8),grid=True)
plt.show()
```
