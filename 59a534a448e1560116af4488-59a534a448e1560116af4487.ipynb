```python
class tfilter():
    def __init__(self, today):
        self.today = today
        self.stocks= DataAPI.EquGet(equTypeCD=u"A", field=u"ticker").ticker.tolist()

    def filter_mvp_2daysIn(self):
        dates = util.get_last_tradeday('', self.today, 3)
        fa = DataAPI.MktEquFlowOrderGet(ticker=self.stocks,beginDate=dates[0],endDate=dates[-1])
        f2 = fa[(fa.tradeDate==dates[0]) & (fa.netInflowXl<0)]
        f1 = fa[(fa.tradeDate==dates[1]) & (fa.netInflowXl<0) & (fa.ticker.isin(f2.ticker))]
        f0 = fa[(fa.tradeDate==dates[-1]) & (fa.netInflowXl>0) & (fa.ticker.isin(f1.ticker))]
        bb = DataAPI.MktRANKInstTrGet(tradeDate=self.today, field=u"ticker", pandas="1")
        ##ff = f0[f0.ticker.isin(bb.ticker)]
        return f0.ticker.values.tolist()
    
    def filter_mvp_inout(self, test_stocks):
        fa = DataAPI.MktEquFlowOrderGet(ticker=self.stocks,beginDate=self.today,endDate=self.today)
        mvp_in = fa[(fa.netInflowXl>0) & (fa.ticker.isin(test_stocks))].ticker.values.tolist()
        mvp_out = fa[(fa.netInflowXl<0) & (fa.ticker.isin(test_stocks))].ticker.values.tolist()
        return mvp_in, mvp_out


from datetime import datetime,timedelta
class util():
    @staticmethod
    def get_last_tradeday(begindate, enddate, count=None):
        if count == None:
            df = DataAPI.TradeCalGet(exchangeCD=u"XSHE", beginDate=begindate, endDate=enddate,field=u"calendarDate,isOpen")
            return df[df['isOpen']==1].calendarDate.values.tolist()
        else:
            t = timedelta(days=20+(count*2))
            n = datetime.strptime(enddate, "%Y%m%d")
            df = DataAPI.TradeCalGet(exchangeCD=u"XSHE", beginDate=n-t, endDate=enddate,field=u"calendarDate,isOpen")
            return df[df['isOpen']==1].calendarDate.values.tolist()[-count:]
                

```
