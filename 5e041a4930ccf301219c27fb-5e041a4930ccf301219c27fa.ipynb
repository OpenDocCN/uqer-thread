```python
import seaborn
from matplotlib import pylab
import matplotlib.pyplot as plt
from matplotlib.pyplot import *
from CAL.PyCAL import *    # CAL.PyCAL中包含font
from datetime import timedelta,datetime
import pandas as pd
import lib.dateHelp as app
import matplotlib.ticker as ticker

#筛选高管增持股票
# beginDate=u"20191201"

# endDate=app.preDay()
# 昨天
beginDate=app.get_date_str(6)
# 今天
endDate=app.get_date_str()


print(u"开始日期-结束日期=("+beginDate+","+endDate+")")

# exchangeCD	calendarDate	isOpen	prevTradeDate	isWeekEnd	isMonthEnd	isQuarterEnd	isYearEnd
#1. 获取交易日历
dateList=DataAPI.TradeCalGet(exchangeCD=u"XSHG",beginDate=beginDate,endDate=endDate,isOpen=u"1",
                             field=u"calendarDate,prevTradeDate",pandas="1")

# 股票池
# universe=set_universe("HS300")+set_universe("ZZ500")
universe=set_universe("A")
# 获取持仓变动情况
raw_df=DataAPI.EquMsChangesGet(secID=universe,ticker=u"",
                        beginDate=beginDate,endDate=endDate,fillDateBegin=u"",fillDateEnd=u"",
                        field=u"ticker,changeReason,changeDate,shareChanges,avgPrice",pandas="1")
raw_df['change_money']=raw_df['shareChanges'] * raw_df['avgPrice']
raw_df["change_money"]=raw_df['change_money'].fillna(0).astype(int)
del raw_df['shareChanges']
del raw_df['avgPrice']
raw_df=raw_df.groupby(["changeDate","ticker"])["change_money"].sum()
raw_df=pd.DataFrame(raw_df,columns=["change_money"]).reset_index()

#raw_df.rename(columns = {"changeDate": "date"},  inplace=True)
raw_df = raw_df[['change_money','ticker', 'changeDate']]
raw_df.index = pd.to_datetime(raw_df['changeDate'])
#raw_df = raw_df[(raw_df.index<=endDate)&(raw_df.index>=beginDate)]  #分析这个时间段之间的高管增持数据

#获取股票名称
data=DataAPI.EquGet(secID=u"",ticker=raw_df["ticker"].tolist(),equTypeCD=u"A",listStatusCD=u"",exchangeCD="",field=u"",pandas="1")
name_dict=data.set_index("ticker")["secShortName"].to_dict()
raw_df["secShortName"]=raw_df["ticker"].map(name_dict)

#最后一个交易日增持的股票池
# raw_df[raw_df.changeDate==dateList["calendarDate"].iloc[-1]]
# 获取上一个交易日增持300w的股票
# raw_df[(raw_df["change_money"]>=3000000) | (raw_df["change_money"]<=-3000000)]

raw_df=raw_df[(raw_df["change_money"]>=3000000) | (raw_df["change_money"]<=-3000000)]

def changeMoreThan300w(money):
    """
    计算买卖是否多于300万
    """
    if money>=3000000:
        return 1
    elif money<=-3000000:
        return -1
    return 0
if len(raw_df):
    raw_df["buy"]=(raw_df["change_money"]).apply(lambda x : 1 if x>=3000000 else 0)
    raw_df["sale"]=(raw_df["change_money"]).apply(lambda x : 1 if x<=-3000000 else 0)
    print(raw_df.to_html())
    print(raw_df[["secShortName","buy","sale"]].groupby(by=["secShortName"]).sum().to_html())
```
