# 基于历史状态空间相似性匹配的行业配置SMIA 模型

**行业轮动分析**
====================
本文主要参考广发金工的研报[《基于历史状态空间相似性匹配的行业配置SMIA 模型》](http://www.doc88.com/p-998527416215.html)中所用的方法，通过量化实验室平台完成实证的过程。

1.**策略思想**
-----------------
随机过程可以划分为有记忆性特征和无记忆性特征两种。行业轮动具有记忆性，即当期行业的相对表现会影响下一期行业的相对表现，这是由行业之间错综复杂的经济关系所决定的。我们通过寻找历史状态空间中与当期行业收益率排名相似的一些时间点，观察这些时间点之后一期行业轮动的变化特征，从中选取统计上表现较好的行业，作为当前时间点下一期的推荐超配行业，从而实现相似性匹配行业配置（Similarity Matching Industry Allocation,缩写为SMIA）量化模型的构建。

2.**实证过程**
-----------------


```python
from CAL.PyCAL import *
import numpy as np
import pandas as pd
from pandas import DataFrame 
from heapq import nlargest
```

```python
def getIndustryInfo(date='20150101'):
    # 这里做简单的处理，暂不考虑不同年份行业的各种变化，实际上这个API取得的数据保留了许多现在已经退市的股票，所以直接使用对实证影响不大。
    # 得到的数据存在许多包含股票数很少的行业，并且当今已经退市，因为不是特别清楚这个API的处理方式，这里将股票数小于10的行业删除。
    # 得到的股票数据还包含许多B股，我们只取A股。
    
    df_ind = DataAPI.EquIndustryGet(industryVersionCD="010303",intoDate=date, field=['secID','industryName1'])
    indName = df_ind['industryName1'].drop_duplicates() #将所有行业汇总
    
    #构建一个{行业：[所包含股票]}的字典
    indContent = {} 
    for ind_name in indName:
        tmp = df_ind[df_ind['industryName1']==ind_name]['secID'].tolist() 
        if len(tmp) > 10:
            indContent[ind_name] = [s for s in tmp if s[0] in '036']
            
    return indContent

# 输出行业股票个数
indContent = getIndustryInfo()
for k,v in indContent.items():
        print '{0}: {1}'.format(k, len(v))
```

```python
def universeGet(indContent):
    # 根据行业的各股票设置股票池
    universe_all = []
    for key in indContent:
        universe_all = universe_all + indContent[key]      
    return universe_all

x = universeGet(getIndustryInfo())
print len(x)
x[:20]
```

```python
def indYieldMeanGet(indContent,yieldSeries):
    #根据股票行业分类以及日期序列，计算各期行业平均收益率
    Ind_meanYield = {}
    closePrice, secReturn = DataFrame(), DataFrame()
    for ind_Name in indContent:
        for i in xrange(len(yieldSeries)):
            trade_date = yieldSeries[i].strftime('%Y%m%d')
            tmp1 = DataAPI.MktEqudAdjGet(secID=indContent[ind_Name], tradeDate=trade_date, field=['closePrice'])
            closePrice = pd.concat([closePrice,tmp1],axis=1) #得到某一个行业的月度股票价格×各股票二维数据框
        
        for i in xrange(len(yieldSeries)-1):
            tmp2 = closePrice.ix[:,i+1]/closePrice.ix[:,i]-1 #一行代表一个股票的月度收益率
            secReturn = pd.concat([secReturn,tmp2],axis=1) 
        
        Ind_meanYield[ind_Name] = secReturn.mean(axis=0).values #用行业内所有股票的平均收益率代表行业收益率
        closePrice, secReturn = DataFrame(), DataFrame()
    return Ind_meanYield
```

```python
#得到80期(每22个工作日为一期)各行业平均收益率
#运算很慢，请自带瓜子

def fixYield(): 
    #得到股票行业分类
    indContent = getIndustryInfo()
    
    #得到月度(每22个工作日)序列
    cal = Calendar('China.SSE')
    end = Date(2013,12,31)
    yieldSeries = [end]
    
    #选取前80期作为数据源
    for i in xrange(80): 
        yieldSeries.append(cal.advanceDate(yieldSeries[-1], '-22B', BizDayConvention.Preceding))
    yieldSeries = yieldSeries[::-1]
    
    fixInd_meanYield = indYieldMeanGet(indContent,yieldSeries)
    return fixInd_meanYield

fixInd_meanYield = fixYield()
DataFrame(fixInd_meanYield).head(8)
```

```python
#得到调仓日到2014-01-02，各期行业平均收益率
def addYield(current_date):
    #得到股票行业分类
    indContent = getIndustryInfo()
    
    #得到月度(每22个工作日)序列
    start = Date(2014,1,2)
    cal = Calendar('China.SSE') #上海交易所工作日日历
    timeLength = cal.bizDatesList(start, Date.fromDateTime(current_date)) #得到effectiveDate与current_date之间的工作日list
    step = len(timeLength) / 22 #判断过了几期
    yieldSeries = [start]
    for i in xrange(step):
        yieldSeries.append(cal.advanceDate(yieldSeries[-1], '22B', BizDayConvention.Preceding)) #得到月度(22B)序列
    addInd_meanYield = indYieldMeanGet(indContent,yieldSeries)
    return addInd_meanYield
```

```python
def distanceGet(history, test):
    #计算欧式距离
    distance = sum((history - test).values**2)
    return distance
```
