# 动量策略-上证50

```python
#对上证50使用动量策略，其中不断的调整前期涨跌幅的区间长度
start = "2014-01-01"
end = "2018-5-31"
benchmark = 'HS300' 
universe = set_universe("HS50")
capital_base = 10000000

# 把回测参数封装到SimulationParameters 中，供quick_backtest 使用
sim_params = quartz.SimulationParameters(start, end, benchmark, universe, capital_base)

# # 获取回测行情数据.idxmp是
# idxmap, data = quartz.get_backtest_data(sim_params)
# # 运行结果，定义为空字符串

results = {}

for x in [5,10,20,60]:#分别测试5天、10天、20天、60天均线的回测情况
    # -----------策略逻辑部分开始，可编辑------------
    # refresh_rate 的值和回测行情数据无关，所以也可以在参数调优中修改
    # 类似的还有commission 和slippage
    refresh_rate = 1
    # commission = ...
    # slippage = ...
    def initialize(account):#定义初始化函数
        account.my_paramter = x
        account.buylist=[ ]
        return 
        
    def handle_data(account):#定义核心逻辑函数
        hist = account.get_attribute_history("closePrice", x)
        for s in account.universe:
            if hist[s][-1]/hist[s][0]-1 > 0.05 and s not in account.valid_secpos:
                order(s, 10000)
            elif hist[s][-1]/hist[s][0]-1 < 0 and s in account.valid_secpos:
                order_to(s, 0)
         
    
    # 把回测逻辑封装到TradingStrategy 中，供quick_backtest 使用
    strategy = quartz.TradingStrategy(initialize, handle_data)
    bt, acct = quartz.quick_backtest(sim_params,strategy,idxmap,data,refresh_rate = 1)
    # 对于回测的结果，可以通过perf_parse 函数计算风险指标
    perf = quartz.perf_parse(bt, acct)#perf为字典类型
    # 保存运行结果
    results[x] = perf["returns"]
    
# 直接打印
print(results)
# 转换为DataFrame 并画图
import pandas

results = pandas.DataFrame(results)
results.plot()
```
