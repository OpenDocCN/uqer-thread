# 证券化率

```python
import pandas as pd
gdp=pd.read_csv('gdp.csv')
del gdp['Unnamed: 0']
```

```python
date=DataAPI.TradeCalGet(exchangeCD=u"XSHG",beginDate=u"20050101",endDate=u"20171111",field=u"calendarDate,isOpen",pandas="1")
date=date[date.isOpen==1]
date.index=range(len(date))
```

```python
# 得到每月最后一个交易日
month=[]
tem=1
month_end=[]
for i in range(len(date)):
    if month==[]:
        month=date.calendarDate.ix[i][:7]
    tem=date.calendarDate.ix[i][:7]
    if tem==month:
        continue
    else:
        month=date.calendarDate.ix[i][:7]
        month_end.append(date.calendarDate.ix[i-1])
        
#month_end

# 修改gdp日期的格式。
for i in range(len(gdp)):
    gdp.quarter.ix[i]=str(gdp.quarter.ix[i]).replace('.1','-03-31').replace('.2','-06-30').replace('.3','-09-30').replace('.4','-12-31')
```

```python
# 得到市值和gdp的表
df=pd.DataFrame(columns=['neg_mv','mv','gdp','Index'], index=month_end) # mv is market value, neg_mv is circulation market value.
pool=set_universe('A')
for t in month_end:
    mv=DataAPI.MktEqudGet(tradeDate=t,secID=pool,ticker=u"",beginDate=u"",endDate=u"",isOpen="",field=u"",pandas="1")
    s_mv=mv.marketValue.sum()/100000000
    s_negmv=mv.negMarketValue.sum()/100000000
    df.mv.ix[t]=s_mv       # 得到股票总市值
    df.neg_mv.ix[t]=s_negmv    # 得到股票总流通市值
    # 得到GDP_ttm
    new_gdp=gdp[gdp.quarter<t][0:5]
    if new_gdp.quarter.ix[new_gdp.index[0]][5:7] =='12' :
        df.gdp.ix[t]=new_gdp.gdp[new_gdp.index[0]]
    if new_gdp.quarter.ix[new_gdp.index[0]][5:7] =='09' :
        df.gdp.ix[t]=new_gdp.gdp[new_gdp.index[0]]+new_gdp.gdp[new_gdp.index[-2]]-new_gdp.gdp[new_gdp.index[-1]]
    if new_gdp.quarter.ix[new_gdp.index[0]][5:7] =='06' :
        df.gdp.ix[t]=new_gdp.gdp[new_gdp.index[0]]+new_gdp.gdp[new_gdp.index[-3]]-new_gdp.gdp[new_gdp.index[-1]]
    if new_gdp.quarter.ix[new_gdp.index[0]][5:7] =='03' :
        df.gdp.ix[t]=new_gdp.gdp[new_gdp.index[0]]+new_gdp.gdp[new_gdp.index[-4]]-new_gdp.gdp[new_gdp.index[-1]]    
    # 得到上证指数收盘价
    df.Index.ix[t]=DataAPI.MktIdxdGet(tradeDate=t,indexID=u"",ticker=u"000001",exchangeCD=u"XSHE,XSHG",field=u"indexID,ticker,secShortName,closeIndex",pandas="1").closeIndex.ix[0]
```

```python
df['sec_rate']=df.mv/df.gdp
df['cir_sec_rate']=df.neg_mv/df.gdp
# df.to_csv('sec_rate.csv',encoding='GBK')
```
