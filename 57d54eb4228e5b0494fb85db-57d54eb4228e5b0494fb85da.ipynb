###### 2. 选股逻辑：

与配对交易有所相似，先确定个股的关联组合，而后假设个股在一段时间内若价格偏离关联组合，则会有回归的倾向。
  即：在过去一段时间内低于关联组合价格的个股未来表现较好，反之未来表现较差（注意：这里的想法本身与“反转”有相似处，只不过其考虑的是相对价格，后者则是绝对价格）
</br>
</br>
###### 3. 选股步骤 (假设月底调仓)：

Step1: 月底确定个股S的关联组合，定义任意两个股票间距离=1-两个股票在过去250个交易日收益率的pearson相关系数。选择距离最小的10只个股作为S的关联组合

Step2: 每天计算关联组合的净值价格，并计算个股S和关联组合的对数价差：PriceSpread=ln(StockPrice−ln(ReferencePrice)

Step3: 调仓日(月底)计算标准化后的PriceSpread，并定义为SpreadBias = (PriceSpread−mean(PriceSpread))/std(PriceSpread)，标准化选样区间为过去60日。

Step4: 超配SpreadBias较低的个股，低配SpreadBias较高的个股
</br>
</br>
###### 4. 方法的探讨，以下程序基本参照原始报告复制，但可能存在的问题是：

(1) 不同于配对交易，此处个股和关联组合之间并没有验证是否真实存在协整关系，即价格的回归只能是一个假设。

(2) 另外，对于PriceSpread时间序列，如果不是I(0)即稳定序列，则计算bias时的标准化操作是有待商榷的
</br>
</br>  
###### 5. 策略构建：

(1) Universe: 沪深300成分股

(2) 时间：2013/02/01-2016/08/31 (新手小白，内存有限，建议多回溯几年更有参考价值**-**)

(3) 选股：按bias从低到高排序，选择top50只股票，等权配置


```python
#### 该模块为一些基础函数
import pandas as pd
import datetime
import time
import numpy as np
import pandas as pd
import math
from CAL.PyCAL import *

# utils函数1：字符和datetime相互转化
def convert_char_to_datetime(chardate, format='%Y-%m-%d',isto=1):
    if isto == 1:
        return datetime.datetime.strptime(chardate,format)
    else:
        return datetime.datetime.strftime(chardate,format)

# utlis函数2：提取最近日期(允许包含或剔除当日) 
def get_nearest_date(charDateList, pivotCharDate, isPrev = 1):
    refDate = [convert_char_to_datetime(x) for x in charDateList]
    pivotDate = convert_char_to_datetime(pivotCharDate)
    if isPrev == 1:
#        refDate = [x for x in refDate if x < pivotDate]
        refDate = filter(lambda x: x < pivotDate, refDate)
        if len(refDate) == 0:
            return ''
        closeDate = min(refDate, key=lambda x: abs(x-pivotDate))
    else:
        refDate = filter(lambda x: x <= pivotDate, refDate)
        closeDate = min(refDate, key=lambda x: abs(x-pivotDate))
    closeDate = convert_char_to_datetime(closeDate,isto=0)
    return closeDate

# utlis函数3：提取A股交易日期
def get_busdate(bdate, edate, exchange=u'XSHG', islist=1):
    caldate = DataAPI.TradeCalGet(exchangeCD=exchange, beginDate=bdate, endDate=edate).sort('calendarDate')
    caldate = caldate[caldate['isOpen'] == 1].calendarDate
    if islist == 1:
        return caldate.values.tolist()
    else:
        return caldate

# utlis函数4：提取各个频率的交易日期
def get_intval_busdate(bdate, edate, exchange=u'XSHG', intchar='W', islist=1):
    caldate = DataAPI.TradeCalGet(exchangeCD=exchange, beginDate=bdate, endDate=edate).sort('calendarDate')
    caldate = caldate[caldate['isOpen'] == 1]

    if intchar.upper() == 'W':
        caldate = caldate[caldate['isWeekEnd'] == 1]
    elif intchar.upper() == 'M':
        caldate = caldate[caldate['isMonthEnd'] == 1]
    elif intchar.upper() == 'Q':
        caldate = caldate[caldate['isQuarterEnd'] == 1]
    elif intchar.upper() == 'Y':
        caldate = caldate[caldate['isYearEnd'] == 1]

    caldate = caldate.calendarDate
    if islist == 1:
        return caldate.values.tolist()
    else:
        return caldate

# 提取A股连续交易日期，并计算最近调仓日(adjDateList)
# 主要用于生成每日的持仓文件
def map_to_adjdate(bdate, edate, adjDateList):
    starttime = datetime.datetime.now()
    dfCaldate = pd.DataFrame({'enddate':get_busdate(bdate=bdate, edate=edate, exchange=u'XSHG', islist=1)})
    refDate = set(adjDateList)
    dfCaldate['adjdate'] = [get_nearest_date(refDate, x) for x in dfCaldate.enddate]
    endtime = datetime.datetime.now()
    print 'The running time of  map_to_adjdate is %f s.' %((endtime - starttime).total_seconds()) 
    return dfCaldate


# 生成每日的持仓文件（该函数后续没有用上，可忽略）
def generate_stock_pool(universe, bdate, edate):
    starttime = datetime.datetime.now()
    dfCaldate = pd.DataFrame({'enddate':get_busdate(bdate=bdate, edate=edate, exchange=u'XSHG', islist=1)})
    dfUniv = pd.DataFrame({'secID': universe})
    dfUniv['key'] = 0
    dfCaldate['key'] = 0
    dfPool = dfUniv.merge(dfCaldate, how='left', on='key' )
    dfPool.index = range(len(dfPool))
    dfPool.drop(['key'], axis=1, inplace=True)
    endtime = datetime.datetime.now()
    print 'The running time of cal generate_stock_pool is %f s.' %((endtime - starttime).total_seconds()) 
    return dfPool


    
# 提取某个交易日的A股名单   
# 留待limit_universe_daily_reference()调用
def get_daily_astock(charDate):
    starttime = datetime.datetime.now()
    aStock = DataAPI.EquGet(equTypeCD=u"A",secID=u"",ticker=u"",listStatusCD=u"",field=u"secID,secShortName,listDate,delistDate,listStatusCD",pandas="1")
    aStock = aStock[(aStock.listDate < charDate) & ((aStock.delistDate > charDate) | (aStock.listStatusCD =='L'))]
    endtime = datetime.datetime.now()
    print 'The running time of get astock pool is %5.5f s.' %((endtime - starttime).total_seconds())
    return aStock.secID

# 提取某日前intval(250)个交易日的行情数据，并计算日收益率（用于之后的相关性矩阵计算）
# 注：将停牌日的收益率0设置为空，以免影响之后相关性计算
# 留待limit_universe_daily_reference()调用
def get_daily_astock_hq(charDate,apool,intval=250):
    starttime = datetime.datetime.now()
    cal = Calendar('China.SSE')
    bintval = Period('-' + str(intval) + 'B')
    bdate =cal.advanceDate(charDate, bintval)
    bdate = bdate.toDateTime().strftime('%Y-%m-%d')
    stockData = DataAPI.MktEqudAdjGet(secID=apool, beginDate=bdate,
                                  endDate=charDate,field=u"secID,tradeDate,closePrice,isOpen",pandas="1")

    stockData['ret'] = stockData['closePrice'].pct_change()
    stockData['previousSec']  = stockData['secID'].shift(1)
    
    stockData.ret[(stockData.secID != stockData.previousSec) | (stockData.isOpen == 0) ] = np.NaN
    stockData.drop(['previousSec','closePrice','isOpen'], axis=1, inplace=True)
    stockData = stockData.set_index(['secID','tradeDate']).unstack('secID')
    stockData.columns = stockData.columns.droplevel(level=0) 
    endtime = datetime.datetime.now()
    print 'The running time of get hq is %5.5f s.' %((endtime - starttime).total_seconds())
    return stockData

# 计算所有A股两两相关性（要求至少有250*0.8=200)个有效值
# 留待limit_universe_daily_reference()调用
def get_daily_reference(charDate,stockData,intval=250,validPct=0.8,refSize=10):
    starttime = datetime.datetime.now()
    minP = math.ceil(intval*validPct)
    stockCorr = stockData.corr(method='pearson',min_periods=minP)
    stockCorr = stockCorr.stack()
    # change different names for reset_index
    stockCorr.index.names=['secID','secID2']
    stockCorr = stockCorr.reset_index()
    stockCorr.columns = ['secID','secID2','corr']
    # filter itself and invalid values
    stockCorr = stockCorr[(stockCorr.secID != stockCorr.secID2) & (pd.notnull(stockCorr.corr))]
    # pick the top N stocks as reference portfolio
    stockCorr.sort_index(by=['secID','corr'], ascending=[True,False], inplace=True)
    stockCorr = stockCorr.groupby('secID').head(refSize)
    stockCorr['enddate'] = charDate 
    endtime = datetime.datetime.now()
    print 'The running time of get reference is %5.5f s.' %((endtime - starttime).total_seconds()) 
    return stockCorr

# 计算universe中，个股的关联组合
def limit_universe_daily_reference(universe,charDate,intval=250,validPct=0.8,refSize=10):
    starttime = datetime.datetime.now()
    apool = get_daily_astock(charDate=charDate)
    stockData = get_daily_astock_hq(charDate=charDate,apool=apool)
    stockCorr = get_daily_reference(charDate=charDate,stockData=stockData,intval=250,validPct=0.8,refSize=10)
    uniCorr = stockCorr[stockCorr.secID.isin(universe.secID)]
    endtime = datetime.datetime.now()
    print 'The running time of get universe reference is %5.5f s.' %((endtime - starttime).total_seconds())     
    return uniCorr

# 逐步计算个股的关联组合，用于get_universe_daily_reference调用
def get_sec_daily_reference(secID,charDate,stockData,intval=250,validPct=0.8,refSize=10):
    starttime = datetime.datetime.now()
    minP = math.ceil(intval*validPct)
    stockCorr = stockData[stockData.columns].apply(lambda x: x.corr(stockData[secID],method='pearson',min_periods=minP ))
    stockCorr = stockCorr[pd.notnull(stockCorr)]
    stockCorr.sort_values(ascending = False, inplace = True)
    stockCorr = stockCorr.head(refSize)
    stockCorr = stockCorr.reset_index()
    stockCorr.columns = ['secID2','corr']
    stockCorr['secID'] = secID
    stockCorr['enddate'] = charDate
    endtime = datetime.datetime.now()
    print 'The running time of get reference is %f s.' %((endtime - starttime).total_seconds()) 
    return stockCorr

# 计算universe中，个股的关联组
# 作为：limit_universe_daily_reference()的另外一种实现方式（当时只是为了比较效率..)
def get_universe_daily_reference(universe, charDate, intval=250, validPct=0.8, refSize=10):
    starttime = datetime.datetime.now()
    uniCorr = pd.DataFrame()
    apool = get_daily_astock(charDate=charDate)
    stockData = get_daily_astock_hq(charDate=charDate,apool=apool)
    
    for i in range(0,len(universe.secID)):
        secID = universe.secID[i]
        print 'The %d records: %s' %(i, secID)
        stockCorr = get_sec_daily_reference(secID=secID,charDate=charDate,stockData=stockData)
        uniCorr = uniCorr.append(stockCorr, ignore_index = True)
    endtime = datetime.datetime.now()
    print 'The running time of get universe reference is %f s.' %((endtime - starttime).total_seconds()) 
    return uniCorr


# 计算每日个股及其对应关联组合的净值，并计算每日的spread
def cal_ref_intval_price(dfPool, dfMapdate, dateUniCorr):
    # 针对每日持仓文件(dfPool)，映射关联组合调仓日
    dfPriceRaw = dfPool.merge(dfMapdate, how='left', on='enddate')
    
    # 映射个股当前的关联组合
    dfPrice = dfPriceRaw.merge(dateUniCorr, how='left', left_on = ['secID', 'adjdate'], right_on = ['secID','enddate'])
    dfPrice.rename(columns={'enddate_x':'enddate'}, inplace=True)
    dfPrice.drop(['enddate_y'], axis=1, inplace=True)
    
    # 提取行情数据（包含关联组合中的成分股和universe中的个股）
    mindate1 = min([convert_char_to_datetime(x) for x in filter(lambda x: x!='', dfPrice.adjdate)])
    mindate2 = min([convert_char_to_datetime(x) for x in filter(lambda x: x!='', dfPrice.enddate)])
    mindate = convert_char_to_datetime(min(mindate1, mindate2), isto=0)
    
    maxdate = max([convert_char_to_datetime(x) for x in filter(lambda x: x!='', dfPrice.enddate)])
    maxdate = convert_char_to_datetime(maxdate, isto=0)
    universe = list(set(filter(lambda x: pd.notnull(x), dfPrice.secID))|set(filter(lambda x: pd.notnull(x), dfPrice.secID2)))
    stockData = DataAPI.MktEqudAdjGet(secID=universe, beginDate=mindate, endDate=maxdate,field=u"secID,tradeDate,closePrice,isOpen",pandas="1")
    
    # 计算个股当前日期和调仓日期之间的净值(merge itself) 
    dfRes1 = dfPriceRaw.merge(stockData, how='left',left_on=['secID', 'enddate'],right_on=['secID','tradeDate'])
    dfRes1.drop(['tradeDate'], axis=1, inplace=True)
    dfRes1.rename(columns={"closePrice":"close","isOpen":"isTrade"}, inplace=True)

    dfRes1 = dfRes1.merge(stockData, how='left',left_on=['secID', 'adjdate'],right_on=['secID','tradeDate'])
    dfRes1.drop(['tradeDate','isOpen'], axis=1, inplace=True)
    dfRes1.rename(columns={"closePrice":"adjClose"}, inplace=True)
    dfRes1['netValue'] = dfRes1['close']/dfRes1['adjClose']

    # 计算关联组合当前日期和调仓日期之间的净值(merge reference)
    dfRes2 = dfPrice.merge(stockData, how='left',left_on=['secID2', 'enddate'],right_on=['secID','tradeDate'])
    dfRes2.drop(['tradeDate','secID_y'], axis=1, inplace=True)
    dfRes2.rename(columns={"closePrice":"refClose",'isOpen':"refIsTrade","secID_x":"secID"}, inplace=True)

    dfRes2 = dfRes2.merge(stockData, how='left',left_on=['secID2', 'adjdate'],right_on=['secID','tradeDate'])
    dfRes2.drop(['tradeDate','secID_y','isOpen'], axis=1, inplace=True)
    dfRes2.rename(columns={"closePrice":"refAdjClose","secID_x":"secID"}, inplace=True)

    # calculate net value
    dfRes2['refNetValue'] = dfRes2['refClose']/dfRes2['refAdjClose']

    grouped = dfRes2.groupby(['secID','enddate'], as_index=False)
    dfRes2 = grouped.agg({'secID2': lambda x: pd.notnull(x).sum(),
                'refIsTrade':lambda x: sum(x==1),
                'refNetValue':'mean'})
    # just for practice of grouped.apply
    # tt2 = grouped.apply(lambda x: x[pd.notnull(x['secID2'])]['refNetValue'].mean())
    dfRes2.rename(columns={"secID2":"nRef"}, inplace=True)
    
    # 计算价差spread
    dfRes = dfRes1.merge(dfRes2, how='left', on=['secID', 'enddate'])
    dfRes['spread'] = np.log(dfRes['netValue']) - np.log(dfRes['refNetValue'])
    refValue = dfRes[['enddate','adjdate','secID','isTrade','refIsTrade','netValue','refNetValue','nRef','spread']]
    return refValue


# 计算个股和关联组合之间的bias
def cal_bias(charDate, universe, refValue, intval=60, refMin=8):
    starttime = datetime.datetime.now()
    cal = Calendar('China.SSE')
    bintval = Period('-' + str(intval) + 'B')
    bdate =cal.advanceDate(charDate, bintval)
    bdate = bdate.toDateTime().strftime('%Y-%m-%d')
    
    ref = refValue[[convert_char_to_datetime(bdate) < convert_char_to_datetime(x) <= convert_char_to_datetime(charDate) for x in refValue.enddate]]
    # not filter
    grouped = ref.groupby(['secID'], as_index=False)
    res1 = grouped['spread'].agg({'nspread': lambda x: pd.notnull(x).sum(),
                    'spreadMean': 'mean',
                    'spreadStd': 'std'})
    # filter
    grouped = ref[ref['nRef']>=refMin].groupby(['secID'], as_index=False)
    res2 = grouped['spread'].agg({'nspreadFilter': lambda x: pd.notnull(x).sum(),
                    'spreadMeanFilter': 'mean',
                    'spreadStdFilter': 'std'})
    bias = refValue[refValue.enddate==charDate]
    bias = bias.merge(res1, how='left', on='secID').merge(res2, how='left', on='secID')
    bias['bias'] = (bias['spread']-bias['spreadMean'])/bias['spreadStd']
    bias['biasFilter'] = (bias['spread']-bias['spreadMeanFilter'])/bias['spreadStdFilter']
    bias[bias.nRef<refMin]['biasFilter'] = np.NaN
    bias = bias[['enddate','secID','isTrade','nRef','spread',
                'nspread','spreadMean','spreadStd',
               'nspreadFilter','spreadMeanFilter','spreadStdFilter',
               'bias','biasFilter']]
    return bias
```

```python
#### 该模块创建一些最终存储变量（可考虑写入文件或数据库等）

# 创建database
# (1) DBuniverse: 每天的Universe(daily)
# (2) DBrefport：每次调仓后的reference portfolio(period, i.e. Month End)
# (3) DBrefvalue：每天的reference portfolio value(daily)
# (4) DBbias：每次调仓后的bias(period, i.e. Month End)

DBuniverse = pd.DataFrame(columns=['enddate','secID'])
DBrefport = pd.DataFrame(columns=['enddate','secID','secID2','corr']) 
DBrefvalue = pd.DataFrame(columns=['enddate','adjdate','secID','isTrade','refIsTrade','netValue','refNetValue','nRef','spread'])
DBbias = pd.DataFrame(columns=['enddate','secID','isTrade','nRef','spread',
                'nspread','spreadMean','spreadStd',
               'nspreadFilter','spreadMeanFilter','spreadStdFilter',
               'bias','biasFilter'])
```

```python
### 该模块模拟回溯区间内，每日相关信号文件的创建

# 确定调仓和关联组合重定义的频率都为月频率(MonthEnd)
envBdate = '2000-01-01'
envEdate = '2016-08-31'
monthEnd = get_intval_busdate(bdate=envBdate, edate=envEdate, exchange=u'XSHG', intchar='M', islist=1)

# 回溯信号区间
testBdate = '2012-12-31'
testEdate = '2016-08-31'
dailyList = get_intval_busdate(bdate=testBdate, edate=testEdate, exchange=u'XSHG', intchar='D', islist=1)

# Step1：创建universe，定义为沪深300成分股（实时）
def loop_create_universe(charDate):
    starttime = datetime.datetime.now()
    universe = pd.DataFrame()
    universe['secID'] = DataAPI.IdxConsGet(ticker='000300',intoDate=charDate,isNew=u"",field=u"consID",pandas="1").consID.tolist()
    universe['enddate'] = charDate
    universe = universe[['enddate','secID']]
    universe.sort_values(by=['enddate','secID'], ascending=True, inplace=True)
    endtime = datetime.datetime.now()
    print 'The running time of loop_create_universe is %f s.' %((endtime - starttime).total_seconds()) 
    return universe

# Step2: 创建关联组合文件(period)
def loop_create_ref_port(charDate, universe):
    starttime = datetime.datetime.now()
    uniCorr = limit_universe_daily_reference(universe=universe, charDate=charDate)
    # alternative
    # uniCorr = get_universe_daily_reference(universe=universe, charDate=charDate)
    uniCorr = uniCorr[['enddate','secID','secID2','corr']]
    endtime = datetime.datetime.now()
    print 'The running time of loop_create_ref_port is %f s.' %((endtime - starttime).total_seconds()) 
    return uniCorr

# Step3: 计算关联组合净值和spread(daily)
def loop_cal_ref_value(charDate, universe):
    starttime = datetime.datetime.now()
    dailyMap = map_to_adjdate(bdate=charDate, edate=charDate, adjDateList=monthEnd)
    dfRes = cal_ref_intval_price(dfPool=universe, dfMapdate=dailyMap, dateUniCorr=DBrefport)
    endtime = datetime.datetime.now()
    print 'The running time of loop_cal_ref_value is %f s.' %((endtime - starttime).total_seconds()) 
    return dfRes

# Step4: 计算universe中个股的bias(period)
def loop_cal_bias(charDate, universe):
    starttime = datetime.datetime.now()
    bias = cal_bias(charDate=charDate, universe=universe, refValue=DBrefvalue)
    endtime = datetime.datetime.now()
    print 'The running time of loop_cal_bias is %f s.' %((endtime - starttime).total_seconds()) 
    return bias

# 模拟每日的操作，调用以上4个函数，并写入数据库(模拟)中
for charDate in dailyList:
    print 'Current Date: ', charDate
    # Step1: (daily)generate universe
    universe = loop_create_universe(charDate=charDate)
    # append to DBuniverse
    DBuniverse = DBuniverse[DBuniverse.enddate != charDate]
    DBuniverse = DBuniverse.append(universe, ignore_index=True)

    # Step2: (conditional)if month end, generate reference portfolio
    if charDate in monthEnd:
        uniCorr = loop_create_ref_port(charDate=charDate, universe=universe)
        # append to DBrefport
        DBrefport = DBrefport[DBrefport.enddate != charDate]
        DBrefport = DBrefport.append(uniCorr, ignore_index=True)

    # Step3: (daily)generate stock pool
    refValue = loop_cal_ref_value(charDate=charDate, universe=universe)
    DBrefvalue = DBrefvalue[DBrefvalue.enddate != charDate]
    DBrefvalue = DBrefvalue.append(refValue, ignore_index=True)
    
    # Step4: (conditional)if month end, calculate bias
    if charDate in monthEnd:
        bias = loop_cal_bias(charDate=charDate, universe=universe)
        # append to DBrefport
        DBbias = DBbias[DBbias.enddate != charDate]
        DBbias = DBbias.append(bias, ignore_index=True)
```

1. 完备的多因子策略本应该遵循一些前序的分析，包括但不限于：因子IC计算/分组表现及多空组合分析/新因子与其他基础因子关系分析等，而后在进行策略模拟。但此处跳过这些分析直接策略，有偷懒的嫌疑...

2. 从因子构建逻辑来看，该因子更可能与“反转”类因子有高相关性。这点有待查验

3. 策略逻辑稳健性的探讨可考虑：调整关联组合定义，来查验是否依旧成立

4. 好像策略的运行时间消耗还是比较大的。简单测试了下，瓶颈主要有两个:

* 提取A股250天的行情以用于计算相关性矩阵这步骤，平均是8-10s左右。

* 计算所有A股两两相关性矩阵的时候，平均是25-30s。这里尝试了用循环的方式，只针对universe中的个股，逐次计算其与其他A股的相关性。但每个计算一次也需要0.6-0.7s，这样子对于universe(=HS300)这样内有300只股票的，也需要18-21s，并没有经济太多。如果换个universe，例如ZZ500，耗时就超过30s。还不如一次性把全A都算了，然后在limit to universe。这里计算相关性矩阵主要用的是df.corr()的方法，可能还有其他效率更高的算法（有知道的小伙伴也请告知，谢谢~）

Anyway，这两步合起来大概就需要30-40s，每个月底关联组合重定义就会碰到。其他的日常交易日算算关联组合净值什么的，看了下基本是0.6-1s内搞定。

因此回溯一年下来，保守估计最多大概需要的总时长就是40s*12(Month)+1s*250(days)=730s~=12min。

因为内存的限制，然后也有点懒得等(希望1H搞定)，这里就挑了2013年开始，沪深300基本也算经历了一个震荡-牛-熊的全周期切换了。有兴趣的自然也可以做做其他的universe或者拉长回溯区间。
