# Notebook无法打开

已经保存的 Notebook，如果出现无法打开的情况。可以根据如下步骤进行排查。

- 检查空间是否已满。

	1. 在左下角的系统状态栏，点击 **内存占用XX%**，打开系统状态详情。
	2. 观察 **存储** 状态条，当前占用的存储空间是多少。
	3. 如果发现已占用的存储空间已经超过90%，则可能是由于 **存储空间已满** 的原因造成的。可以查看本文下方 “1. 内存已满的解决办法” 的内容，尝试解决。

- 检查是不是所有的Notebook都无法打开。

	1. 点击其他 Notebook，查看是否可以正常打开。
	2. 如果只有当前 Notebook 无法打开，则表明仅是该 Notebook 的问题，可能是 Notebook 损坏，或者内容过大。可以查看本文下方 “2. Notebook 内容过大的解决办法”，尝试解决。

- 检查是否是 Notebook 内容过大。
	1. 清空 Notebook 的全部输出内容。清空办法见 “2. Notebook 内容过大的解决办法”。
	2. 如果清空输出后，依旧无法打开，可能是 Notebook 损坏。可以参见 “3. Notebook 损坏的解决办法”。

# Notebook无法打开的解决办法

完成上述的检查，确定问题后，可以使用如下办法，解决对应的问题。

## 1. 存储空间已满的解决办法

由于每个优矿用户的存储空间都是有上限的，所以如果存储空间已满，比较实际的解决办法是删除不需要的数据文件。
操作流程如下：

- 通过导航栏的 **数据**，进入数据列表，
 
  ![图片注释](http://storage-uqer.datayes.com/57c4f3f1228e5b6d277c77c3/75235118-1856-11e8-85a4-0242ac140002)

- 删除不常用的数据文件。可以在删除前下载进行备份。

  ![图片注释](http://storage-uqer.datayes.com/57c4f3f1228e5b6d277c77c3/791d23d4-1856-11e8-85a4-0242ac140002)

---
## 2. Notebook 内容过大的解决办法

Notebook 包含代码部分和运算结果输出部分。如果运算结果输出部分过大，会造成整个 Notebook 文件的大小过大。
操作流程如下：

- 点击 Notebook 右侧的下拉列表。点击 “清除所有单元的输出”

  ![图片注释](http://storage-uqer.datayes.com/57c4f3f1228e5b6d277c77c3/d2c2482e-1856-11e8-85a4-0242ac140002)

---
## 3. Notebook 损坏的解决办法

当 Notebook 损坏，如果您有将该 Notebook 进行备份，则可以通过历史版本恢复到之前版本的内容。
操作流程如下：

- 点击页面右上角导航栏的 **时钟** 图标。

  ![图片注释](http://storage-uqer.datayes.com/57c4f3f1228e5b6d277c77c3/4bc670ec-1857-11e8-927b-0242ac140002)


- 查看是否有历史版本，如果有历史版本，则点击 **“查看此版本”**。

  ![图片注释](http://storage-uqer.datayes.com/57c4f3f1228e5b6d277c77c3/4ef1a142-1857-11e8-85a4-0242ac140002)


- 如果该版本为您需要的内容，则点击 **“恢复当前版本”**。

  ![图片注释](http://storage-uqer.datayes.com/57c4f3f1228e5b6d277c77c3/513d4da2-1857-11e8-85a4-0242ac140002)
