```python
############################################
### decorators
############################################

def wrapper_func_cache(func):
    @wraps(func)
    def wrapped(*args, **kwargs):
        if not hasattr(func, '__cache_result'):
            func.__cache_result = {}
        
        hash_key = [str(i) for i in args] + [str(j) for j in kwargs.iteritems()]
        cache_key = hash(tuple(hash_key))
        
        if cache_key not in func.__cache_result:
            func.__cache_result[cache_key] = func(*args, **kwargs)
            
        return func.__cache_result[cache_key]
    return wrapped

def wrapper_func_time(func):
    @wraps(func)
    def wrapped(*args, **kwargs):
        tstart = time.time()
        res = func(*args, **kwargs)
        tend = time.time()
        
        print '-' * 80
        print 'consume {} seconds for func {}'.format(tend - tstart, func.func_name)
        return res
    return wrapped

def wrapper_func_map(map_name):
    def wrapper(func):
        global HELPER_MAP
        if map_name not in HELPER_MAP:
            HELPER_MAP[map_name] = Map()

        HELPER_MAP[map_name][func.func_name] = func
        return func
    return wrapper
```

```python
############################################
### graph
############################################

def fig_style(ax, legend, legend_loc='upper right', xlabel='', grid=True):
    ax.legend(legend, prop=small_font, loc=legend_loc)
    ax.title.set_font_properties(font)
    if xlabel:
        ax.set_xlabel(xlabel, fontproperties=small_font)
    ax.set_axis_bgcolor('white')
    if grid:
        ax.grid(color='gray', alpha=0.2, axis='y')
    return ax
```

```python
############################################
### constants
############################################

data = DataAPI.SecIDGet(assetClass=u"E", field=['ticker', 'secShortName'])
const_ticker_name_dict = {i: j for i, j in zip(data.ticker, data.secShortName)}
const_ticker_name_dict.update({j: i for i, j in zip(data.ticker, data.secShortName)})

@wrapper_func_cache
def const_trade_calendar():
    data = DataAPI.TradeCalGet(exchangeCD=u"XSHG,XSHE", beginDate=u"20050101", endDate=u"", field=['calendarDate', 'isOpen'])
    data = data[data.isOpen == 1]

    return list(data.calendarDate)

@wrapper_func_cache
def const_pre_trade_day(day):
    trade_days = const_trade_calendar() 
    index = trade_days.index(day)
    
    return trade_days[index - 1]
```

```python
############################################
### trade day related
############################################

def trade_range(tstart, tend):
    data = DataAPI.TradeCalGet(exchangeCD=u"XSHG", beginDate=tstart, endDate=tend, field=['calendarDate', 'isOpen'])
    return list(data[data.isOpen == 1]['calendarDate'])
    
def trade_pre_n_days(tend='', ndays=1):
    trade_days = const_trade_calendar() 
    tend = tend if tend else dt.datetime.today().strftime("%Y-%m-%d")
    
    while tend not in trade_days:
        tend = (dt.datetime.strptime(tend, '%Y-%m-%d') - dt.timedelta(days=1)).strftime('%Y-%m-%d')
        
    index = trade_days.index(tend)    
    return trade_days[index - ndays : index + 1]

@wrapper_func_cache
def is_trade_day(day):
    if isinstance(day, (str, unicode)):
        pass
    elif isinstance(day, dt.datetime):
        day = day.strftime("%Y-%m-%d")
    trade_days = const_trade_calendar()
    
    return day in trade_days

def latest_trade_day():
    """最近两个交易日
    """
    today = dt.datetime.today().strftime('%Y-%m-%d')
    trade_days = const_trade_calendar()
    while today not in trade_days:
        today = dt.datetime.strptime(today, '%Y-%m-%d') - dt.timedelta(days=1)
        today = today.strftime("%Y-%m-%d")
        
    pre_trade_day = const_pre_trade_day(today)
    
    return pre_trade_day, today
```

```python
############################################
### utils
############################################
@wrapper_func_cache
def get_all_ticker(tradeDate=None, preTradeDate=None, sector=['主板', '中小板', '创业板']): 
    if not tradeDate or not preTradeDate:
        preTradeDate, tradeDate = latest_trade_day()
    tmp = DataAPI.EquGet(equTypeCD="A", field=['ticker', 'secID', "listDate", "ListSector", "totalShares"])
    tmp = tmp[tmp.ListSector.isin(sector)]
    equ_pool = DataAPI.MktEqudGet(tradeDate=preTradeDate,
                                  field=["ticker", "tradeDate", "exchangeCD", "secShortName","preClosePrice", "openPrice", 
                                         "dealAmount", "turnoverValue","highestPrice", "lowestPrice", "closePrice", 
                                         "negMarketValue", "marketValue", "isOpen"])
    tmp = pd.merge(left=equ_pool, right=tmp, left_on="ticker", right_on="ticker")
    
    ### 过滤掉预退市股
    tmp = tmp[~tmp.secShortName.str.contains('退市')]
    return tmp
```
