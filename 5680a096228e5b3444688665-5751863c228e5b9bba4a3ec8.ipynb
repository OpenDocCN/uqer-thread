# 【大师系列】-16-理查马文价值导向

## 理查和马文是价值投资的拥护者，在投资组合选股中，尤其重视分红率高且稳定的公司。二人于 1980 年代开始合作研究关系，曾先后于核心财务公司（ CoreStates Financial）、伯恩财务集团合作进行投资方面的研究。之后二人正式开始以价值投资方式进行投资组合管理，其投资业绩始终保持在同类型资产管理公司的前列。
