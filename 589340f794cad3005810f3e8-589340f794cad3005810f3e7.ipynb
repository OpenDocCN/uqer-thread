pandas的dataframe中有三个常用的合并工具：1）merge；2）concat；3）combine_first（这个不是亲爹养的）

```python
import pandas as pd
import numpy as np
```

# merge

```python
dt1 = pd.DataFrame({'a':{'11':1},'b':{'22':3},'c':{'33':9}})
dt1.ffill(inplace=True)

dt2 = pd.DataFrame({'a':{'21':1},'e':{'22':4},'f':{'33':9}})
dt2.fillna(0,inplace=True)

print dt1
print dt2
```

```python
dt1.merge(dt2)
```

默认情况下的merge合并是inner，下面详谈merge中比较重要的四个参数：1）how；2）on；3）suffixes；4）left_index/right_index；

```python
dt1.merge(dt2,how='outer',on=None,suffixes=('_left','_right'),left_index=True,right_index=True)
```

how有四种，分别是outer/inner/left/right。首先outer对应inner，内连接和外连接的区别形象地说，**一个是拼图，一个是剪纸**，拼图是完整的两个表格凑在一起，剪纸么……表格就要修要改；left对应right，这俩是一个玩意儿，左连接和右连接，那他们连接的左右是什么列呢？看下面例子:

```python
dt_left = dt1.merge(dt2,how='left')
dt_right = dt1.merge(dt2,how='right')
print dt_left
print dt_right
```

很清楚，左（右）代表着左（右）边dataframe的元素，选哪边，哪边就是老大，凡是异类统统去除。

suffixes上面的例子已经很清楚了，想花式改名的可以随便改……只要不改变元祖结构，这里要重点说说on

```python
print dt1.merge(dt2,how='inner',on='a',suffixes=('_left','_right'),left_index=True,right_index=True)
print dt1.merge(dt2,how='outer',on='a',suffixes=('_left','_right'),left_index=True,right_index=True)
```

这里的on指定合成列，df1和df2共有的键，和outer/inner配合完成,right_on和left_on也是选键的，一个左dataframe的，一个右dataframe的；

```python
print dt1.merge(dt2,suffixes=('_left','_right'),left_index=True,right_index=True)
print dt1.merge(dt2,suffixes=('_left','_right'))
```

最后有个left_index/right_index，根据index合并,哪个TRUE跟哪个

# concat

- 有了merge，为什么还要concat呢？恩……金融序列专用VIP工具

concat专业从事索引合并工作，看下面例子：

```python
pd.concat([dt1,dt2])#这里对象是dataframe的list或者一个字典（或series）
```

```python
pd.concat([dt1,dt2],axis=1)#换个方向合并
```

```python
pd.concat([dt1,dt2],axis=0,join='inner')#换成inner，默认outer（这里跟merge不同）
```

```python
pd.concat([dt1,dt2],keys=['A','B'],names=['hello','bye'])#keys赋予层级索引，顺便用name赋值
```

```python
pd.concat([dt1,dt2],ignore_index=True)#去除index序列
```

# combine_first

- 为什么还会有combine_first？这个东西是一个便捷的合并方法，目的是当两个dataframe按照索引合并时，如果元素为NAN，则自动填入有值的那个dataframe元素（默认优选dt1）

```python
dt1.combine_first(dt2)
```
