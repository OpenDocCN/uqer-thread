```python
##股票选取与处理
def chs_ticker(ticker = '000005',begin = '20010101',end = '20170201'):
    dataday = DataAPI.MktEqudGet(tradeDate=u"",secID=u"",ticker=ticker,beginDate= begin,endDate= end,isOpen="1",field=u'',pandas="1")
    k = list(pd.qcut(dataday['chgPct'],4,labels = ['1','2','3','4']))
    l = k[0]
    for i in k:
        l = l+i
    l = l[1:]
    return l
```

```python
l = chs_ticker('600759')
```

```python
##序列生成
def n_list(glist = l,x = 2):
    relist = []
    for k in range(0,len(glist)-x+1):
        relist.append(glist[k:k+x])
    return relist
```

```python
##支持度计算
l = chs_ticker('600759')
relist = n_list(l,2)
total_num = len(relist)
i = 0
frq = []
sup = []
for k in set(relist):
    if relist.count(k)/np.float(total_num)>(1/np.float(4**2+2**2)):
        frq.append(k)
        sup.append(relist.count(k)/np.float(total_num))
```

```python
##ck序列与置信度生成函数
def ck(frq,x):
    klist = []
    frq_k = []
    sup_k = []
    bay_s = []
    re_list = n_list(l,x-1)
    ren_list = n_list(l,x)
    total_num = len(ren_list)
    for i in frq:
        b = np.float(re_list.count(i))
        for j in ['1','2','3','4']:
            k = i+j
            p = ren_list.count(k)
            if p/np.float(total_num)>0.01:
                if (p/np.float(total_num))>(1/np.float(4**x+2**x)):
                    frq_k.append(k)
                    sup_k.append(p/np.float(total_num))
                    bay_s.append(p/b)
    return frq_k,sup_k,bay_s
```
