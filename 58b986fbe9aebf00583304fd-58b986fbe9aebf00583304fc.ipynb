# 1、MA均线择时策略
* 本策略是采用长短均线金叉买入，死叉卖出方法；
* 参数已经进行了一定的优化，还应该有进一步优化的空间。

```python
from __future__ import division
import pandas as pd
import numpy as np
from CAL.PyCAL import *    # CAL.PyCAL中包含font
import datetime
import talib as ta

def MA_pd(to_day,short=22,long=25):
    dp=0
    nn = '-'+str(max(short,long)+1)+'B'
    cal = Calendar('CHINA.SSE')	# 股票交易所日历
    last_day = cal.advanceDate(to_day,'-1B',BizDayConvention.Preceding).strftime("%Y-%m-%d")
    last_n_day = cal.advanceDate(to_day,nn,BizDayConvention.Preceding).strftime("%Y-%m-%d")
    hs300_close = DataAPI.MktIdxdGet(tradeDate=u"",indexID=u"",ticker=u"000300",beginDate=last_n_day,endDate=last_day,exchangeCD=u"XSHE,XSHG",field=u"tradeDate,closeIndex",pandas="1")
    ma_short = hs300_close.closeIndex.values[-short:].mean()
    ma_short_last= hs300_close.closeIndex.values[-short-1:-1].mean()
    ma_long = hs300_close.closeIndex.values[-long:].mean()
    ma_long_last = hs300_close.closeIndex.values[-long-1:-1].mean()
    if ma_short>ma_long and ma_short_last<=ma_long_last:
        dp=1
    elif ma_short<ma_long and ma_short_last>=ma_long_last:
        dp=-1
    return dp
```
