# 银行股轮动策略

在[集思录](http://www.jisilu.cn/question/50176)上看到有人用**翻倍期**这个指标进行银行股轮动，在帖子中定义了几种翻倍期的计算方法，先用优矿的因子API写个最基本的。

下面是**资产翻倍期**在帖子中的定义：

> * 评价指标很多有市盈率、市净率、资产收益率等等。如何综合量化是个难题。这里我提出我“资产翻倍期”的观点。即假定公司不分红，资产收益率不变，每年的收益利滚利，达到股价两倍所需要的时间。 在EXCEL表中用函数:LOG(市净率两倍，资产收益率+1)，翻倍期越小越好。


```python
import numpy as np
from math import log

def get_all_bank_secID_and_tickers():
    """获取所有银行股的的secID和ticker"""
    sw_industry = DataAPI.EquIndustryGet(industry=u"",secID=u"",ticker=u"",industryVersionCD=u"010303",intoDate=u"",field=u"",pandas="1")
    
    # 480101 代表银行类代码
    bank_df = sw_industry[(sw_industry["industrySymbol"]==480101) & (sw_industry["isNew"]==1)]
    return bank_df['secID'].tolist(), bank_df['ticker'].tolist()


def get_pb_and_roe(date_str,tickers):
    return DataAPI.MktStockFactorsOneDayGet(tradeDate=date_str,secID=u"",ticker=tickers,field=u"secID,ticker,ROE,PB",pandas="1")


def get_pre_trading_day(date_str):
    trading_day_df = DataAPI.TradeCalGet(exchangeCD=u"XSHG",beginDate=date_str,endDate=date_str,field=u"",pandas="1")
    prevTradeDate = trading_day_df["prevTradeDate"].iloc[0]
    return prevTradeDate[0:4] + prevTradeDate[5:7] + prevTradeDate[8:]   # change time formate

    
```
