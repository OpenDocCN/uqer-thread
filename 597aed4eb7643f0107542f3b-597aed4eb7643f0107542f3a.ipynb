```python
# Define Functions
from cvxopt import matrix,solvers
def MinVarPortfolio(sig2,constraints = {}):
    hlen = sig2.shape[0]
    G,h,A,b = encode_constraints(np.zeros(hlen),sig2,constraints)
    sol = solvers.qp(P = matrix(sig2), q = matrix(np.zeros((hlen,1))),
                     A = A, b = b, G = G, h = h)
    if sol['status'] == 'optimal':
        res = np.array(sol['x']).flatten()
    else:
        raise Exception('Failed.')
    return res

def encode_constraints(mu,sig2,constraints):
    hlen = len(mu)
    A = matrix(np.ones((1,hlen)))
    b = matrix(np.array([[1.0]]).T)
    G,h = np.array([]).reshape((0,hlen)),np.array([]).reshape((0,1))
    if constraints is None:
        G,h = None,None
    else:
        for key in constraints.keys():
            if key == 'minW':
                G = np.row_stack((G,-np.eye(hlen)))
                h = np.row_stack((h,-np.ones((hlen,1))*constraints['minW'].reshape((-1,1))))
            elif key == 'maxW':
                G = np.row_stack((G,np.eye(hlen)))
                h = np.row_stack((h,np.ones((hlen,1))*constraints['maxW'].reshape((-1,1))))
            elif key == 'minsumW':
                pairs,cons = constraints['minsumW']
                for p,c in zip(pairs,cons):
                    temp = np.zeros((1,hlen))
                    temp[0,p] = 1
                    G = np.row_stack((G,-temp))
                    h = np.row_stack((h,-c))
            elif key == 'maxsumW':
                pairs,cons = constraints['maxsumW']
                for p,c in zip(pairs,cons):
                    temp = np.zeros((1,hlen))
                    temp[0,p] = 1
                    G = np.row_stack((G,temp))
                    h = np.row_stack((h,c))
        G,h = matrix(G),matrix(h)
    return G,h,A,b

import scipy.stats as sps
def linear(x,y,sigma_flag = False,mean = True,rd = 0):
    
    X = np.column_stack((np.repeat(1,len(x)),x))
    if not mean:
        X = X[:,1:]
    Y = y.reshape((-1,1))
    temp_X = np.linalg.inv(X.T.dot(X) + rd * np.eye(X.shape[1]))
    beta = temp_X.dot(X.T).dot(Y)
    res = Y - X.dot(beta)
    sigma = temp_X*np.std(res)**2
    sigma = np.array([np.sqrt(sigma[i,i]) for i in range(len(sigma))])
    t_values = beta.flatten()/sigma
    p_values = 2*(1-sps.t(len(X)-len(beta)).cdf(np.abs(t_values)))
    p_values = np.round(p_values,4)
    R2 = 1 - np.sum(res**2)/np.sum((Y - np.mean(Y))**2)
    
    if sigma_flag:
        return beta.flatten(),sigma,R2,res
        
    return beta.flatten(),p_values,R2,res

def step_foward(x,y,num,fields = None):
    k = 1
    res = y
    X = []
    locs = []
    while k <= num:
        r = np.abs([np.cov(x[:,i],res.flatten())[0,1]/\
                    (np.std(x[:,i])*np.std(res)) for i in range(x.shape[1])])
        loc = np.argmax(r)
        X.append(x[:,loc])
        x = x[:,np.array(range(x.shape[1]))!=loc]
        xt = np.array(X).T
        try:
            beta, p_values,R2,res = linear(xt,y)
        except:
            beta, p_values,R2,res = linear(xt,y,rd = 0.0001)
        k += 1
        if fields is not None:
            locs.append(fields[loc])
            fields = fields[np.array(range(len(fields)))!=loc]
    if fields is None:
        return np.array(X).T
    else:
        return np.array(X).T,locs

def RobustAdaboost(X,Y,QN = 5,L = 100,tol = 1e-6,mode = 'Classifier',criteria = 'Discrimination',fcycle = 0):
    
    hlen = X.shape[0]
    vlen = X.shape[1]
    eps = 1.0/hlen
    
    # Initialize Sample Weights
    Weights = np.repeat(1.0/hlen,hlen)
    
    # Divided Quantiles
    Qs = np.array([np.percentile(X,q,axis = 0) for q in np.linspace(0,100,QN + 1)[1:]])
    
    # Best Factors
    BestFactors = np.array(np.zeros(L),dtype = int)
    BestWsPositive = np.zeros((L,QN))
    BestWsNegative = np.zeros((L,QN))
    BestWs = np.zeros((L,QN))
    WeakRegressorWeights = np.zeros(L)
    old_BestFactors = None
    old_BestFactors_list = []
    for i in range(L):
        SignedWeight = (Y*Weights).reshape((-1,1))
        Ws_positive = np.zeros((len(Qs),vlen))
        Ws_negative = np.zeros((len(Qs),vlen))
        Ws = np.zeros((len(Qs),vlen))
        Yps = np.zeros((hlen,vlen))
        for j in range(len(Qs)):
            if j == 0:
                loc = (X <= Qs[j]) 
            else:
                loc = (X <= Qs[j]) & (X > Qs[j-1])
            if mode == 'Regressor':
                Quantiled_Weight = Weights.reshape((-1,1))*loc
                Quantiled_Weight_normed = Quantiled_Weight / (np.sum(Quantiled_Weight, axis = 0) + 1e-14)
                Ws[j] = np.sum(Quantiled_Weight_normed*Y.reshape((-1,1)),axis = 0)
                WeakScores = Ws[j]
            else:
                Quantiled_SignedWeight = SignedWeight*loc
                positive_loc = Quantiled_SignedWeight > 0
                negative_loc = Quantiled_SignedWeight < 0
                Ws_positive[j] = np.sum(Quantiled_SignedWeight*positive_loc,axis = 0)
                Ws_negative[j] = -np.sum(Quantiled_SignedWeight*negative_loc,axis = 0)
                WeakScores = 1.0/2.0*np.log((Ws_positive[j] + eps)/(Ws_negative[j] + eps))
            Yps += WeakScores*loc

        if mode == 'Regressor':
            if criteria == 'Discrimination':
                Zs = np.zeros(vlen)
                for zloc in range(vlen):
                    Zs[zloc] = np.sum(np.abs(Ws[:,zloc].reshape((-1,1)) - Ws[:,zloc]))
                BestFactors[i] = np.argmax(Zs)
                while old_BestFactors is not None and BestFactors[i] in old_BestFactors_list:
                    Zs[BestFactors[i]] = np.min(Zs)
                    BestFactors[i] = np.argmax(Zs)
            else:
                Zs = np.sum(np.abs(Y.reshape((-1,1)) - Yps)*Weights.reshape((-1,1)),axis = 0)
                BestFactors[i] = np.argmin(Zs)
                while old_BestFactors is not None and BestFactors[i] in old_BestFactors_list:
                    Zs[BestFactors[i]] = np.max(Zs)
                    BestFactors[i] = np.argmin(Zs)
            BestWs[i] = Ws[:,BestFactors[i]]
            Error = np.abs(Y - Yps[:,BestFactors[i]])
            MaxError = np.max(Error)
            LinearError = Error/MaxError
            TotalErrorRate = np.sum(Weights*LinearError)
            WeakRegressorWeight = TotalErrorRate/(1 - TotalErrorRate)
            WeakRegressorWeights[i] = WeakRegressorWeight
            Weights *= WeakRegressorWeight**(1 - LinearError)
            Weights /= np.sum(Weights)
            if WeakRegressorWeight > 1 - tol:
                break
        else:
            Zs = np.sum(np.sqrt(Ws_positive*Ws_negative),axis = 0)
            BestFactors[i] = np.argmin(Zs)
            while old_BestFactors is not None and BestFactors[i] in old_BestFactors_list:
                Zs[BestFactors[i]] = np.max(Zs)
                BestFactors[i] = np.argmin(Zs)
            BestWsPositive[i] = Ws_positive[:,BestFactors[i]]
            BestWsNegative[i] = Ws_negative[:,BestFactors[i]]
            Weights *= np.exp(-Y*Yps[:,BestFactors[i]])
            Weights /= np.sum(Weights)

        if i != 0 and np.sum(np.abs(Yps[:,BestFactors[i]])) < tol and BestFactors[i] == old_BestFactors:
            break
        old_BestFactors = BestFactors[i]
        old_BestFactors_list.append(BestFactors[i])
        if len(old_BestFactors_list) > fcycle:
            old_BestFactors_list.pop(0)
        
    if mode == 'Regressor':
        return BestFactors[:i],BestWs[:i],WeakRegressorWeights[:i],Qs,eps
    else:
        return BestFactors[:i],BestWsPositive[:i],BestWsNegative[:i],Qs,eps

def RobustAdaboostPredict(X,model,mode = 'Classifier'):
    
    if mode == 'Regressor':
        BestFactors,BestWs,WeakRegressorWeights,Qs,eps = model
    else:
        BestFactors,BestWsPositive,BestWsNegative,Qs,eps = model
    hlen = len(X)
    Fp = X[:,BestFactors]
    Qsp = Qs[:,BestFactors]
    Yps = np.zeros((hlen,len(BestFactors)))
    for j in range(len(Qs)):
        if j == 0:
            loc = (Fp < Qsp[j]) 
        else:
            loc = (Fp < Qsp[j]) & (Fp > Qsp[j-1])
        if mode == 'Regressor':
            WeakScores = BestWs[:,j]
        else:
            WeakScores = 1.0/2.0*np.log((BestWsPositive[:,j] + eps)/(BestWsNegative[:,j] + eps))
        Yps += WeakScores*loc
    if mode == 'Regressor':
        Yp = Yps.dot(np.log(1/WeakRegressorWeights))/np.sum(np.log(1/WeakRegressorWeights))
    else:
        Yp = Yps.sum(axis = 1)
    
    return Yp
```

```python
# Get Month Data
dates = DataAPI.MktIdxdGet(indexID = '000001.ZICN', beginDate = '20130101', endDate = '20170630', field = 'tradeDate')['tradeDate'].values
dates = np.array(dates, dtype = 'datetime64[D]')
dates_M = np.array(dates, dtype = 'datetime64[M]')
month_unique = np.unique(dates_M)
month_start_end = []
for month in month_unique:
    temp_dates = dates[dates_M == month]
    month_start_end.append([str(temp_dates[0]),str(temp_dates[-1])])
```

```python
# Get Realized Month Return
RealizedRets = []
for row in month_start_end:
    startDate = row[0].replace('-','')
    endDate = row[1].replace('-','')
    startPrice = DataAPI.MktEqudAdjGet(secID = universe, tradeDate = startDate, field = ['secID','tradeDate','closePrice'])
    endPrice = DataAPI.MktEqudAdjGet(secID = universe, tradeDate = endDate, field = ['secID','tradeDate','closePrice'])
    startIndex = startPrice['secID'].values
    startPrice.index = startIndex
    endIndex = endPrice['secID'].values
    endPrice.index = endIndex
    endPrice = endPrice.reindex(startIndex)
    assert all(endPrice['closePrice'].values > 0)
    RealizedRet = endPrice['closePrice'] / startPrice['closePrice'] - 1
    RealizedRets.append(RealizedRet)
    print(row[1] + ' ok')
```

```python
# Get Factors
MonthFactors = []
for row in month_start_end:
    endDate = row[1]
    date = endDate
    temp_factors = DataAPI.MktStockFactorsOneDayGet(tradeDate = date, secID = universe,field = factor_list, pandas="1")
    temp_factors = temp_factors.dropna()
    temp_factors = temp_factors.set_index(temp_factors['secID'])
    date_t = date.replace('-', '')
    for factor in factor_list[2:]:
        #temp_factors[factor] = pd.Series(standardize(neutralize(winsorize(temp_factors[factor].to_dict(),win_type='QuantileDraw', pvalue=0.05),date_t, risk_module='short', industry_type='SW2')))
        temp_factors[factor] = pd.Series(standardize(winsorize(temp_factors[factor].to_dict(),win_type='QuantileDraw', pvalue=0.05)))
    tol_factors = temp_factors

    factors = tol_factors[factor_list[2:]].values*1
    del_loc = []
    for i in range(len(factors)):
        if np.sum(np.abs(factors[i,:])) < 1e-5:
            del_loc.append(i)
    in_loc = list(set(np.arange(len(factors))) - set(del_loc))
    factors = factors[in_loc]
    factors_index = tol_factors['secID'].values[in_loc]
    factors = pd.DataFrame(factors)
    factors.index = factors_index
    
    MonthFactors.append(factors)
    
    print(date + ' ok')
```

```python
# Train Model

import sklearn.linear_model as sl
import sklearn.ensemble as se
models = []
factor_locs = []
XTolTrain,YTolTrain = None,None
train_window = 8
train_sample_lens = []
for factors,RealizedRet,row in zip(MonthFactors[:-1],RealizedRets[1:],month_start_end[1:]):
    X = factors.values.argsort(axis = 0).argsort(axis = 0) / float(len(factors))
    Y = RealizedRet.reindex(factors.index).values
    assert all(~np.isnan(Y))
    #locs = step_foward(X,Y,10,factors.columns)[1]
    locs = np.arange(factors.values.shape[1])
    X = factors.loc[:,locs].values.argsort(axis = 0).argsort(axis = 0) / float(len(factors))
    factor_locs.append(locs)
    Y_temp = Y.argsort().argsort() / float(len(Y)) # rank
    X_temp = X
    
    # Clasifier
    #Y_temp_loc = (Y_temp < 0.3) | (Y_temp > 0.7)
    #Y_temp = Y_temp[Y_temp_loc]
    #Y_temp[Y_temp < 0.3] = -1
    #Y_temp[Y_temp > 0.7] = 1
    #X_temp = X[Y_temp_loc,:]
    
    train_sample_lens.append(len(X_temp))

    if XTolTrain is None:
        XTolTrain,YTolTrain = X_temp,Y_temp
    else:
        XTolTrain,YTolTrain = np.row_stack((XTolTrain,X_temp)),np.concatenate((YTolTrain,Y_temp))
        
    if len(train_sample_lens) > train_window:
        pop_pos = train_sample_lens.pop(0)
        XTolTrain,YTolTrain = XTolTrain[pop_pos:],YTolTrain[pop_pos:]

    #model = sl.LinearRegression()
    #model.fit(XTolTrain,YTolTrain)
    #yp = model.decision_function(X)
    
    #model = se.RandomForestRegressor(n_estimators = 20,max_depth = 1,random_state = 20)
    #model.fit(XTolTrain,YTolTrain)
    #yp = model.predict(X)
    
    model = RobustAdaboost(XTolTrain,YTolTrain,5,100,1e-4,mode = 'Regressor',fcycle = 0)
    yp = RobustAdaboostPredict(X,model,mode = 'Regressor')
    
    yt = Y.flatten()
    R2 = 1 - np.sum((yp - yt)**2)/np.sum((yt - np.mean(yt))**2)
    models.append(model)
    # Rank R2
    yp_rk = yp.argsort().argsort()
    yt_rk = yt.argsort().argsort()
    np.percentile(yt_rk[yp_rk > 100],50)
    # Output
    hlen = len(yt_rk)
    winrate = float(np.sum(yt_rk[yp_rk > hlen - 50] > hlen / 2))/len(yt_rk[yp_rk > hlen - 50])
    mean_return = np.mean(yt[yp_rk > hlen - 50])
    expected_rank = np.mean(yt_rk[yp_rk > hlen - 50])
    print('TimeRange: %s, Winrate: %s, MeanReturn: %s, ERank: %s, SampleNum: %s'%(row[0] + ' to ' + row[1],round(winrate,4),round(mean_return,4),round(expected_rank,2),hlen))
    
```

```python
# Validation and Test
tol_ranks = []
winrates = []
portfolioReturns = []
benchmarkReturns = []
stockCodes = []
basket_num = 40
for factors,RealizedRet,row,model,locs in zip(MonthFactors[1:-1],RealizedRets[2:],month_start_end[2:],models[:-1],factor_locs[:-1]):
    X = factors.loc[:,locs].values.argsort(axis = 0).argsort(axis = 0) / float(len(factors))
    Y = RealizedRet.reindex(factors.index).values
    assert all(~np.isnan(Y))
    yp = RobustAdaboostPredict(X,model,mode = 'Regressor')
    #yp = model.predict(X)
    #yp = model.decision_function(X)
    yt = Y.flatten()
    R2 = 1 - np.sum((yp - yt)**2)/np.sum((yt - np.mean(yt))**2)
    # Rank R2
    yp_rk = yp.argsort().argsort()
    yt_rk = yt.argsort().argsort()
    # np.percentile(yt_rk[yp_rk > 100],50)
    # Output
    hlen = len(yt_rk)
    winrate = float(np.sum(yt_rk[yp_rk > hlen - basket_num] > hlen / 2))/len(yt_rk[yp_rk > hlen - basket_num])
    winrates.append(winrate)
    mean_return = np.mean(yt[yp_rk > hlen - basket_num])
    mean_return_total = np.mean(yt)
    expected_rank = np.mean(yt_rk[yp_rk > hlen - basket_num])
    tol_ranks.extend((yt_rk[yp_rk > hlen - basket_num]/float(hlen)).tolist())
    # Record Returns
    portfolioReturns.append(mean_return)
    benchmarkReturns.append(mean_return_total)
    # Record StockCodes
    stockCodes.append(np.array(factors.index)[yp_rk > hlen - basket_num])
    print('TimeRange: %s, Winrate: %s, MeanReturn: %s, MeanReturnTotal: %s, ERank: %s, SampleNum: %s'%(row[0] + ' to ' + row[1],round(winrate,4),round(mean_return,4),round(mean_return_total,4),round(expected_rank,2),hlen))
print('tol_percentile50_rank: %s'%(np.percentile(tol_ranks,50)))
print('WinRate: %s'%(np.sum(np.array(tol_ranks)>0.5)/float(len(tol_ranks))))
```

```python
# Performance Parameters

def MDB(data):
    hlen = len(data)
    Max = data[0]
    res = []
    for i in range(1,hlen):
        if data[i] > Max:
            Max = data[i]
        res.append((Max - data[i])/Max)
    return np.max(res)

def CAGR(data):
    hlen = len(data)
    res = (data[-1]/data[0])**(12.0/hlen)-1
    return res

portfolioEquity = np.cumprod(1+np.array(portfolioReturns))
benchmarkEquity = np.cumprod(1+np.array(benchmarkReturns))
abnormalEquity = np.cumprod(1 + np.array(portfolioReturns) - np.array(benchmarkReturns))
print('策略 年化收益： %s, 最大回撤： %s, 收益回撤比： %s'%(CAGR(portfolioEquity),MDB(portfolioEquity),CAGR(portfolioEquity)/MDB(portfolioEquity)))
print('基准 年化收益： %s, 最大回撤： %s, 收益回撤比： %s'%(CAGR(benchmarkEquity),MDB(benchmarkEquity),CAGR(benchmarkEquity)/MDB(benchmarkEquity)))
print('超额 年化收益： %s, 最大回撤： %s, 收益回撤比： %s'%(CAGR(abnormalEquity),MDB(abnormalEquity),CAGR(abnormalEquity)/MDB(abnormalEquity)))
print('选股月度胜率： %s'%(np.sum(np.array(portfolioReturns)>np.array(benchmarkReturns))/float(len(portfolioReturns))))
```
