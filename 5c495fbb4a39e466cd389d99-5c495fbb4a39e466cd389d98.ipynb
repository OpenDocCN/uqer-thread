```python
from __future__ import division
import pandas as pd
import numpy as np
import datetime
from dateutil.parser import parse
from CAL.PyCAL import *
cal=Calendar('China.SSE')
```

计算PEG指标，以及PEG指标和其后收益率的相关系数

首先取得时间截面的选股函数

```python
def selectScotk(universe,date):
    """给定股票列表和日期
    args：
    universe（list or str）：策略中的univer
    date（date or datetime or str）
    returns：
    list：股票列表"""
    trade_date=date if isinstance(date,datetime.datetime) else parse(date)
    trade_date=trade_date.strftime('%Y%m%d')
    benchmarkDate=cal.advanceDate(trade_date,Period('-1Y')).strftime('%Y%m%d')
    EPS_factor=DataAPI.MktStockFactorsOneDayGet(tradeDate=trade_date,secID=universe,ticker=u"",field=u"secID,EPS,PE",pandas="1")
    EPS_factor=EPS_factor.rename(columns={'EPS':'EPS0'})
    EPS_factor1=DataAPI.MktStockFactorsOneDayGet(tradeDate=benchmarkDate,secID=universe,ticker=u"",field=u"secID,EPS",pandas="1")
    EPS_factor1=EPS_factor1.rename(columns={'EPS':'EPS1'})
    EPS_factor=EPS_factor.merge(EPS_factor1,on='secID')
    EPS_factor['g']=EPS_factor['EPS0']/EPS_factor['EPS1']-1
    EPS_factor['PEG']=EPS_factor['PE']/EPS_factor['g']
    EPS_factor=EPS_factor[(EPS_factor['PE']>0)&(EPS_factor['PEG']>0)]
    EPS_factor=EPS_factor.drop(['EPS0','EPS1','PE','g'],axis=1)
    EPS_factor=EPS_factor.sort_values(by='PEG',ascending=True)
    buy_list=EPS_factor.head(10)['secID'].tolist()
    return buy_list
```
