1. 判断日期的方法
----------------------
***

``Python``中的``datetiem``可以帮助判断当前日期是周几：

```python
import datetime as dt
testDate = dt.datetime(2015,5,18)

print testDate.weekday()
```

``weekday``的返回值代表了具体周几。从周一开始到周日分别对应0到6。
