一.获取数据与特征处理

```python
import numpy as np
import pandas as pd

data = DataAPI.MktIdxdGet(indexID=u"",ticker=u"000300",tradeDate=u"",beginDate=u"20100101",endDate=u"20180501",exchangeCD=u"XSHE,XSHG",field=u"",pandas="1")
data.set_index('tradeDate', inplace = True)
# 获取HS300的每日价格数据

for i in range(1, 21, 1):
    data['close-' + str(i) + 'd'] = data['closeIndex'].shift(i)
# 对于收盘价，在data中增加了20列数据，分别为1天前的收盘价至20天前的收盘价
    
hs_close = data[[x for x in data.columns if 'close' in x]].iloc[20:]
# 选取今天及1-20天前的收盘价，iloc[20:]剔除了无效数据

hs_close = hs_close.iloc[:, ::-1]
# 将新DataFrame的列按倒序排列
```

二.定义训练集和测试集、训练模型

PS：运行此代码大约需要5分钟，想缩短时间可以减少数据，并修改days参数。

```python
from sklearn import svm
# 从sklearn库中导入svm算法

days = 1500
# 设定全局变量，分割训练集和测试集的数据，1500在上文数据中约占75%

clf_close = svm.SVR(kernel='linear')
# 使用svm下的SVR算法，'linear'表示线性核
f_close_train = hs_close[:days]
# 训练集features
l_close_train = hs_close['closeIndex'].shift(-1)[:days]
# 训练集labels，将收盘价shift(-1)表示预测的是下一天的收盘价
f_close_test = hs_close[days:]
# 测试集features
l_close_test = hs_close['closeIndex'].shift(-1)[days:]
# 训练集labels，将收盘价shift(-1)表示预测的是下一天的收盘价
clf_close.fit(f_close_train, l_close_train)
# 训练模型
```

三.训练集表现

```python
p_close_train = clf_close.predict(f_close_train)
# 将训练集features导入模型进行预测，生成预测的收盘价
df_close_train = pd.DataFrame(l_close_train)
# 新建一个DataFrame，内容为训练集labels,即下一天的收盘价
df_close_train.columns = ['next close']
# 列名重命名为'next close'
df_close_train['predicted next close'] = p_close_train
# 加入一列预测的收盘价数据
df_close_train['next open'] = data['openIndex'][20:20 + days].shift(-1)
# 加入一列下一天开盘价的数据，从data而非hs_close中获取，需要切片

trigger = 1.0
df_close_train['position'] = np.where(df_close_train['predicted next close'] > df_close_train['next open'] * trigger, 1, 0)
# 通过np.where函数判断，当预测的下一天收盘价 > 下一天开盘价相乘或相加一个trigger时，仓位设置为1，否则为0
df_close_train['PL'] = np.where(df_close_train['position'] == 1, (df_close_train['next close'] - df_close_train['next open']) / df_close_train['next open'], 0)
# 当仓位为1时，在下一天开盘时买入，收盘时卖出，记录下一天应获得的收益率，否则收益率为0

df_close_train['strategy'] = (df_close_train['PL'].shift(1) + 1).cumprod()
# 策略每日收益的累积收益率，其中shift(1)表示当日记录的是下一天才能获得的收益率，当日无法获得
df_close_train['return'] = (df_close_train['next close'].pct_change() + 1).cumprod()
# benchmark的累积收益率

df_close_train[['strategy', 'return']].dropna().plot()
# 画出策略与benchmark的累计收益率图
```

四.测试集表现

不管模型在训练集中表现如何，在测试集中表现好才说明模型有一定的预测性。

```python
p_close_test = clf_close.predict(f_close_test)
# 与训练集代码基本相同，只是将数据换为测试集的数据
df_close_test = pd.DataFrame(l_close_test)
df_close_test.columns = ['next close']
df_close_test['predicted next close'] = p_close_test
df_close_test['next open'] = data['openIndex'][days + 20:].shift(-1)
# 测试集数据与训练集数据切片内容不同

df_close_test['position'] = np.where(df_close_test['predicted next close'] > df_close_test['next open'] * trigger, 1, 0)
df_close_test['PL'] = np.where(df_close_test['position'] == 1, (df_close_test['next close'] - df_close_test['next open']) / df_close_test['next open'], 0)

df_close_test['strategy'] = (df_close_test['PL'].shift(1) + 1).cumprod()
df_close_test['return'] = (df_close_test['next close'].pct_change() + 1).cumprod()

df_close_test[['strategy', 'return']].dropna().plot()
```

五.基础回测指标

构建一个简单的ratio函数来判断收益率、波动率、最大回撤三个指标。

注意：输入的数据为每日收益数据，非累积收益数据。

```python
def ratio(data):
    strategy_cum = (data + 1).cumprod()
    return_year = strategy_cum.tolist()[-1] ** (252./strategy_cum.shape[0]) - 1    
    # 计算收益率
    
    volatility = data.std() * 252 ** 0.5
    # 计算波动率
    
    drawdown = ((strategy_cum.cummax() - strategy_cum)/strategy_cum.cummax()).max()
    # 计算最大回撤
     
    res = pd.DataFrame([return_year, volatility, drawdown], columns=[data.name], index=['return', 'volatility', 'drawdown'])
    # 输出数据为列表[收益率，波动率，最大回撤]， 列名为[data.name]，是取Series下列名的方法， 索引为自定义的['return', 'volatility', 'drawdown']
    return res

result_close = ratio(df_close_test['PL']).join(ratio(df_close_test['next close'].dropna().pct_change()))
# DataFrame下的join方法，按index合并，合并benchmark和策略的回测指标
result_close.columns = ['strategy', 'benchmark']
result_close
```

#### 可以发现，模型在测试集中的表现还不错，在收益率、波动率、最大回撤方面均有提升，尤其是最大回撤这一项很小，可以说是理想的净值曲线了。

六.加入SMA和CCI优化后的特征处理与训练模型

在之前的第三篇文章中，用到了SMA和CCI双指标结合的思路，效果还不错，所以这里尝试使用这两个指标作为特征因子。
其余思路和步骤与上文基本相同，可以套用其他特征因子来进行研究。
PS：运行此代码大约需要20分钟。

```python
import talib as ta
data['sma'] = ta.SMA(np.asarray(data['closeIndex']), 5)
# 获取SMA数据

data['cci'] = ta.CCI(np.asarray(data['highestIndex']), np.asarray(data['lowestIndex']), np.asarray(data['closeIndex']), timeperiod = 21)
# 获取CCI数据

for i in range(1, 21, 1):
    data['sma-' + str(i) + 'd'] = data['sma'].shift(i)
    
hs_sma = data[[x for x in data.columns if 'sma' in x]].iloc[40:]
# 由于CCI数据的特点，有效数据的切片改为[40:]
hs_sma = hs_sma.iloc[:, ::-1]
   
for i in range(1, 21, 1):
    data['cci-' + str(i) + 'd'] = data['cci'].shift(i)
        
hs_cci = data[[x for x in data.columns if 'cci' in x]].iloc[40:]
hs_cci = hs_cci.iloc[:, ::-1]

hs_mix = hs_sma.join(hs_cci)
# 将SMA和CCI的特征因子放入一个DataFrame中

hs_mix.insert(0, 'closeIndex', np.asarray(data['closeIndex'].iloc[40:]))
# 加入所需的labels列

clf_mix = svm.SVR(kernel='linear')
f_mix_train = hs_mix[:days]
l_mix_train = hs_mix['closeIndex'].shift(-1)[:days]

f_mix_test = hs_mix[days:]
l_mix_test = hs_mix['closeIndex'].shift(-1)[days:]
clf_mix.fit(f_mix_train, l_mix_train)
# 训练模型
```

七.训练集表现

这里的代码基本都是复用的，思路和之前是完全一样的。

```python
p_mix_train = clf_mix.predict(f_mix_train)
df_mix_train = pd.DataFrame(l_mix_train)
df_mix_train['predict'] = p_mix_train
df_mix_train['next open'] = data['openIndex'][40:40 + days].shift(-1)
df_mix_train.columns = ['next close', 'predicted next close', 'next open']

df_mix_train['position'] = np.where(df_mix_train['predicted next close'] > df_mix_train['next open'] * trigger, 1, 0)
df_mix_train['PL'] = np.where(df_mix_train['position'] == 1, (df_mix_train['next close'] - df_mix_train['next open']) / df_mix_train['next open'], 0)

df_mix_train['strategy'] = (df_mix_train['PL'].shift(1) + 1).cumprod()
df_mix_train['return'] = (df_mix_train['next close'].pct_change() + 1).cumprod()

df_mix_train[['strategy', 'return']].dropna().plot()
```

八.测试集表现

```python
p_mix_test = clf_mix.predict(f_mix_test)
df_mix_test = pd.DataFrame(l_mix_test)
df_mix_test['predict'] = p_mix_test
df_mix_test['next open'] = data['openIndex'][days + 40:].shift(-1)
df_mix_test.columns = ['next close', 'predicted next close', 'next open']

df_mix_test['position'] = np.where(df_mix_test['predicted next close'] > df_mix_test['next open'] * trigger, 1, 0)
df_mix_test['PL'] = np.where(df_mix_test['position'] == 1, (df_mix_test['next close'] - df_mix_test['next open']) / df_mix_test['next open'], 0)

df_mix_test['strategy'] = (df_mix_test['PL'].shift(1) + 1).cumprod()
df_mix_test['return'] = (df_mix_test['next close'].pct_change() + 1).cumprod()

df_mix_test[['strategy', 'return']].dropna().plot()
```

总结：

- 首先需要注意的是，由于切片的不同，优化后的策略与优化前的策略数据略有不同，所以benchmark的数据会有一些小差异，但无伤大雅。
- 可以看出，优化后的策略在收益率方面有较大提升，但波动率和最大回撤都变大了，毕竟收益和风险难以同时兼顾。
- 粗略计算一下，优化后策略的夏普比率是变大了的，所以总的来说还是达到了优化的目的。
- 策略只做多，且没有考虑交易成本，特征因子的提取也相对简单，这些都是可以改进的方向。
