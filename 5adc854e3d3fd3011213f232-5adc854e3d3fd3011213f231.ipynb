# 2.动量策略与均值回归

### 这次来做一个简单的动量策略和均值回归策略，动量策略与均值回归策略的逻辑看起来刚好相反。动量模型的假设是当前价格是过去价格趋势的延续，而均值回归则认为当前上涨的趋势是由于高估造成的，当前的下跌是由于低估造成的，无论是高估还是低估，都会慢慢的回到均值。我们很难说哪种策略更为正确或者有效，就和经典力学、相对论以及量子力学适用于不同的物理环境，动量策略和均值回归也是类似，适用于不同的市场情况。难点在于，对于动量策略，我们很难预测趋势可以持续多久，对于均值回归，回归需要多长时间同样难以预测。



## 首先来看看动量策略

- 动量（momentum）,有的地方也翻译为惯性，即股票（或投资标的）延续其上涨或下跌趋势的能力。在著名的Fama-French三因子模型中，认为股票的收益率由三大因子决定，即市场因子（market factor）,规模因子（size factor）,价值因子（value factor）。之后Carhart对其进行了改进，加入了一个新的因子--动量因子（momentum factor），即赢家组合与输家组合的差值。研究发现，赢家组合在牛市中存在着正的动量效应，输家组合在熊市中存在着负的动量效应。

- 表示动量的方法有许多种，这里选择最简单的一种方法：当收益率>0时，认为有正的动量，反之则有负的动量。正的动量做多，负的动量做空。

- 与上一篇策略中的双均线SMA类似，动量策略在趋势型市场中效果较好，而在震荡型市场中效果欠佳。

一.数据准备

通过DataAPI.MktIdxdGet接口调取沪深300指数的数据。

```python
import numpy as np
import pandas as pd

data = DataAPI.MktIdxdGet(indexID=u"",ticker=u"000300",tradeDate=u"",beginDate=u"20100401",endDate=u"20180401",exchangeCD=u"XSHE,XSHG",field=u"",pandas="1")
data.set_index('tradeDate', inplace = True)
#将日期设为index

data.head()
```

二.策略思想

pct_change语句可计算每日收益率。
收益率>0时，认为有正的动量，做多，反之做空。
通过np.sign()函数判断收益率的正负，>0时仓位为1，<0时仓位为-1。

```python
data['return'] = data['closeIndex'].pct_change()

data['position'] = np.sign(data['return'])

data.head()
```

三.计算策略收益

将每日收益乘以仓位，即可得到策略收益。
通过对'position'进行shift操作，可避免未来函数。
（收益率 + 1）.cumprod方法计算离散的收益率，上篇文章中SMA的算法是连续的收益率。

```python
data['strategy'] = data['position'].shift(1) * data['return']
#计算策略收益

data['return_cum'] = (data['return'] + 1).cumprod()
data['strategy_cum'] = (data['strategy'] + 1).cumprod()
#计算benchmark和策略的离散累积收益率

data[['return_cum', 'strategy_cum']].dropna().plot()
```

四.基础回测指标构建

构建一个简单的ratio函数来判断收益率、波动率、最大回撤三个指标。

注意：输入的数据为每日收益数据，非累积收益数据。

```python
def ratio(data):
    return_year = data.mean() * 252
    #求收益率
    
    volatility = data.std() * 252 ** 0.5
    #求波动率
    
    strategy_cum = (data + 1).cumprod()
    drawdown = (strategy_cum.cummax() - strategy_cum).max()
    #求最大回撤
     
    res = pd.DataFrame([return_year, volatility, drawdown], columns=[data.name], index=['return', 'volatility', 'drawdown'])
    #输出数据为列表[收益率，波动率，最大回撤]， 列名为[data.name]，是取Series下列名的方法， 索引为自定义的['return', 'volatility', 'drawdown']
    return res

ratio(data['return']).join(ratio(data['strategy']))
#DataFrame下的join方法，按index合并，合并benchmark和策略的回测指标
```

五.策略优化

首先，在之前的策略中，仅通过当天的收益率来判断动量，会造成频繁地开仓买卖，且趋势不一定明显。
所以我们选择观察更长的一段时期，若这段时间内收益率均值>0，则认为正的动量较大，趋势较为明显且减少了买卖次数。

其次，通过简单穷举的办法，找到较为合适的天数作为我们的观察期。
天数的选择不宜过短，否则依旧会频繁买卖：也不宜过长，因为长时期后的动量与当前动量相关性太差。

最终，我们选择5, 10, 15, 20, 30, 60这6个天数作为穷举的代表。


```python
price_plot = ['return_cum']     
#创建一个含有benchmark累积收益率的list

for days in [5, 10, 15, 20, 30, 60]:
    
    data['position_%dd' % days] = np.where(data['return'].rolling(days).mean() > 0, 1, -1)
    #where函数做条件判断，若平均收益率>0则仓位为1，反之为0
    
    data['strategy_%dd' % days] = data['position_%dd' % days].shift(1) * data['return']
    #求策略的每日收益率
    
    data['strategy_cum_%dd' % days] = (data['strategy_%dd' % days] + 1).cumprod()
    #求策略的累积收益率
    
    price_plot.append('strategy_cum_%dd' % days)
    #通过append函数将累积收益率的名称加入之前的list
    
data[price_plot].dropna().plot(figsize = (15, 8))
#画出所有曲线的图像

ratio(data['return']).join(ratio(data['strategy'])).join(ratio(data['strategy_20d']))
#join函数构建最终的DataFrame
```

## 总结：

- 从图中可看出，采用20日的观察期效果最好。而更长时期的效果很差，基本类似于random walk。

- 在没有进行优化前，我们的策略收益率尚可，波动率相差不大，但最大回撤比benchmark要高，这并不是一个理想的策略。
  但经过优化后，不仅收益率获得了较大提升，最大回撤也从0.7370下降到了0.5225，虽然0.52的回撤依旧是很大的数字，但相比之前已经有明显的进步。

- 不足之处：
  没有考虑做空的限制。
  没有考虑交易费用及市场冲击可能产生的滑点。
  未进行样本外检验。

- 动量策略获得了较高的正收益，可见，追涨杀跌未尝不可，但是需要有合适的择时方法。

- 动量策略在趋势型市场中表现较好，接下来我们来研究在震荡型市场中表现较好的策略，也是动量策略的好兄弟----均值回归策略。

## 均值回归策略
- 均值回归（mean reverting），起初是金融学的一个重要概念。均值回归是指股票价格、房产价格等社会现象、自然现象（气温、降水），无论高于或低于价值中枢（或均值）都会以很高的概率向价值中枢回归的趋势。根据这个理论，一种上涨或者下跌的趋势不管其延续的时间多长都不能永远持续下去，最终均值回归的规律一定会出现：涨得太多了，就会向平均值移动下跌；跌得太多了，就会向平均值移动上升。

- 均值回归是价值型投资者的核心思想，以著名的巴菲特为代表。他们判断企业的长期价值，认为股票价格若存在明显的高估或低估的状态，则价格最终会均值回归到价值。CFA体系着重培养价值型投资者，其核心思想便是均值回归。

- 配对交易（pair trading）或协整套利也是以均值回归为核心。当两只股票（标的资产）的时间序列关系满足协整性时，便可以采用配对交易来赚钱。这两只股票的价差（或比值），长期会表现出一种稳定的关系，若某个时刻偏离了这种关系，则认为会出现均值回归的情况，此时便可进行交易。

- 说句题外话，个人认为，我们的世界里最普世的两种现象，应该就是均值回归和正态分布了（不过也有人认为幂律分布更能解释自然界，而正态分布解释人造体的时候效果更好），生活中有非常多的现象与此相关。比方说，我们每个人的朋友，按财富、地位、性格等分类，大概是属于正态分布的。我们平时的状态，考试啊心情啊，大概是遵从均值回归的。anyway，更好地理解这二者，不仅对我们的投资很有帮助，也为我们的生活增添了不少乐趣。

一.数据准备

通过DataAPI.MktIdxdGet接口调取沪深300指数的数据。

```python
import numpy as np
import pandas as pd

data1 = DataAPI.MktIdxdGet(indexID=u"",ticker=u"000300",tradeDate=u"",beginDate=u"20100401",endDate=u"20180401",exchangeCD=u"XSHE,XSHG",field=u"",pandas="1")
data1.set_index('tradeDate', inplace = True)
#将日期设为index

data1.head()
```

二.策略思想

在这个策略中，我们有空仓的情况。
首先计算一段时期内的移动平均值，在这里设为55天。
接着判断当前价格是否严重偏离均值，若价格高于均值，则做空，等待价格回归均值时平仓：若价格低于均值，则做多，等待价格回归均值时平仓。

```python
import matplotlib.pyplot as plt

days = 55
#设定移动窗口

data1['SMA'] = data1['closeIndex'].rolling(days).mean()
#求出移动平均值

data1['diff'] = data1['closeIndex'] - data1['SMA']
# 求出当前价格与移动平均值的差值

threshold = 200
#设定阈值，超出阈值时进行交易

data1['diff'].dropna().plot(legend = True)
plt.axhline(threshold, color = 'r')
plt.axhline(0, color = 'r')
plt.axhline(-threshold, color = 'r')
#画图感受更为直观
```

```python
data1['position'] = np.where(data1['diff'] > threshold, -1, np.nan)
data1['position'] = np.where(data1['diff'] < -threshold, 1, data1['position'])
data1['position'] = np.where(data1['diff'] * data1['diff'].shift(1) < 0, 0, data1['position'])
#当出现上穿或下穿threshold时，产生多空信号：
#data1['diff'] * data1['diff'].shift(1) < 0表示穿过均值，产生平仓信号，继续等待下一个信号的产生

data1['position'] = data1['position'].ffill().fillna(0)
#通过ffill函数，填充了na值，表示之后一段时间内的持仓情况
#fillna(0)表示刚开始时未产生信号，则不持仓

data1['position'].ix[days:].plot(ylim=[-1.1, 1.1], figsize=(10, 6))
#ix[days:]表示剔除了刚开始未产生信号的时期
```

三.计算策略收益

将每日收益乘以仓位，即可得到策略收益。
通过对'position'进行shift操作，可避免未来函数。
（收益率 + 1）.cumprod方法计算离散的收益率，上篇文章中SMA的算法是连续的收益率。

```python
data1['return'] = data1['closeIndex'].pct_change()
#计算每日收益率

data1['strategy'] = data1['position'].shift(1) * data1['return']
#计算策略收益

data1['return_cum'] = (data1['return'] + 1).cumprod()
data1['strategy_cum'] = (data1['strategy'] + 1).cumprod()
#计算benchmark和策略的离散累积收益率

data1[['return_cum', 'strategy_cum']].dropna().plot()
```
