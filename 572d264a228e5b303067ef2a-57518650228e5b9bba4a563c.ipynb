```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame

from datetime import datetime
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#回测区间
start = '20120101'                       # 回测起始时间
end   = '20160414'                         # 回测结束时间

#获取回测期间全部数据
df = DataFrame()
df = DataAPI.MktFunddAdjGet(ticker='510300',beginDate=start,endDate=end,field=u"tradeDate,openPrice,closePrice",pandas="1")
df = df.rename(columns={'tradeDate':'DATE','openPrice':'OPEN','closePrice':'CLOSE'})
df['DATE'] =  pd.to_datetime(df['DATE']) #因为DataAPI所有日期都是以字符串形式存储的，因此此处需要转换成日期时间型,这里很重要！
df = df.set_index('DATE')#转换后index是DatetimeIndex型
```

```python
df.head()
```

1. Let's make new column called SIGNALS, i.e. our 2 bars signal:

```python
returns = df.pct_change().dropna()
#make the signal
returns['SIGNAL'] = (returns['CLOSE'] - returns['OPEN'] + returns.shift()['CLOSE'] - returns.shift()['OPEN'])
```

```python
returns.head()
```

2. Then let's normalize it with round(), meaning we'll be acting to last two bar changes in increments, of, let's say, 1%. In other words, our position size will depend on those previous moves. We already use percentage values, so we just need to round the signal to 2 decimal points. Also we need to shift the signal:

```python
#normalize
returns['NORM'] = returns['SIGNAL'].round(2).shift(2)
```

3. Now let's just get strategy returns. Signal will be generated at last Close and position would be initiated at Open. Also, as our strategy is mean reverting, we just change the sign of the signal:

#signal formed at last close, position opened at Open 
returns['RETURNS'] =  returns['OPEN'] * -returns['NORM']

```python
#Here's what strategy generated over period:
#see signal
plt.plot(returns['NORM'])
plt.show()
```

Visually compared to buy and hold:
We can, for example, leverage our signal, meaning, if signal change was 20%, we would use 200% of capital (also meaning our parameterless strategy has also built-in position sizing):

```python
#signal formed at last close, position opened at Open 
leverage = 10.0
returns['RETURNS'] =  returns['OPEN'] * -returns['NORM']*leverage

#generate strategy curve
plt.plot(returns.CLOSE.cumsum(), color='green')
plt.plot(returns['RETURNS'].cumsum())
plt.show()
```


4. Generate some stats.
We can check the correlation between signal and Close to Close returns:



```python
#check signal correlation with close to close returns
sns.jointplot(x=returns.CLOSE, y=returns.NORM, kind='scatter')
sns.plt.show()
```

We can compare Sharpe ratios:

```python
bhavg = returns['CLOSE'].mean()
stravg = returns['RETURNS'].mean()
print "Average buy and hold return: %s" %bhavg
print "Average strategy return: %s" %stravg

bhstd = returns['CLOSE'].std()
strstd = returns['RETURNS'].std()
print "Buy and hold STD: %s" %bhstd
print "Strategy STD: %s" %strstd

bhsr = bhavg/bhstd*np.sqrt(252)
strsr = stravg/strstd*np.sqrt(252)
print "Buy and hold Sharpe ratio: %s" %bhsr
print "Strategy Sharpe ratio: %s" %strsr
```
