```python
normal_df = pd.pivot_table(acc_nav_df,values='ACCUM_NAV', index='endDate', columns='ticker', aggfunc=np.sum)
print normal_df.columns
```

```python
irr_df = normal_df.pct_change()
```

```python
import math
mean_ser = irr_df.mean() * 252
std_ser = irr_df.std() * math.sqrt(252)
sharpe_ser = mean_ser/std_ser
```

```python
def get_win_ratio(col):
    ix = 0
    while not col[ix] or np.isnan(col[ix]):
        ix += 1
    valid = col.iloc[ix:]
    positive = valid[valid>0]
    return len(positive)/(len(valid)+0.0)
```

```python
win_ser = irr_df.apply(get_win_ratio,reduce=True)
```

```python
 #allocation stage
sort_criteria = {'mean': mean_ser,
            'var': std_ser,
            'sharpe': sharpe_ser,
            'win': win_ser
            }
```

```python
def getPercentAllocation(totalPercent):
    '''greedy allocation'''
    upLimitNumber = int(math.floor(totalPercent/maxAlloc))
    percent_lst = [maxAlloc] * upLimitNumber
    residual = totalPercent - upLimitNumber * maxAlloc        
    if not percent_lst:
        if residual < minAlloc - 1e-4:
            percent_lst = []
        else:
            percent_lst = [residual]
    else:
        if math.fabs(residual) >= 1e-4:
            while residual < minAlloc - 1e-4 :
                percent_lst[-1] -= 0.05
                residual += 0.05
            percent_lst.append(residual)
    return percent_lst       
```

```python

to_sort_ser = sort_criteria[strategy]
print to_sort_ser
to_sort_ser.sort(inplace=True, ascending=(strategy == 'var'))
sorted_ticker_lst = to_sort_ser.index.tolist()
sorted_dic = {}
for kind in category_lst:
    ticker_lst = []
    for ticker in sorted_ticker_lst:
        if ticker in fund_pool[kind]:
            ticker_lst.append(ticker)
    sorted_dic[kind] = ticker_lst
print sorted_dic
```
