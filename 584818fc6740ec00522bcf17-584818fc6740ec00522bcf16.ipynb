```python
import pandas as pd
import numpy as np
import seaborn as sns
sns.set_style('white')
import matplotlib.pyplot as plt
from matplotlib import rc
from matplotlib import dates
rc('mathtext', default='regular')
from scipy.stats import mstats
from scipy import stats
import scipy.stats as st
from datetime import datetime
import time
from CAL.PyCAL import *

# initialization,
begin_date = datetime.strptime('2015-11-01', '%Y-%m-%d')    # initial time
end_date = datetime.strptime('2016-1-1', '%Y-%m-%d')      # end time
univ = set_universe('A') # China A share, 
#univ = ['000001.XSHE', '600000.XSHG'] 

# calculate every stock's forward return and future 'window's days return, 
def getOpenClosePrice(universe, begin, end, file_name):
    print ' open close price for ' + str(len(universe)) + ' stocks:'
    count = 0
    secs_time = 0
    start_time = time.time()#
    
    #data collection
    ret_data = pd.DataFrame()
    N = len(universe)
    
    for stk in universe:
        data = DataAPI.MktEqudAdjGet(secID=stk, beginDate=begin, endDate=end, field='secID,tradeDate,openPrice,closePrice')
        tmp_ret_data = data.sort('tradeDate')

        if ret_data.empty:
            ret_data = tmp_ret_data
        else:
            ret_data = ret_data.merge(tmp_ret_data, on='tradeDate', how='outer')
        
        # print progress
        count += 1
        if count > 0 and count % N == 0:
            finish_time = time.time()
            print count,
            print '  ' + str(np.round((finish_time-start_time) - secs_time, 0)) + ' seconds elapsed.'
            secs_time = (finish_time-start_time)#time count
    
    ret_data.to_csv(file_name)
    return ret_data
    
# calculate future 10 days of return,               
print '======================='
start_time = time.time()
getOpenClosePrice(universe=univ, begin=begin_date, end=end_date, file_name='openclose_price_test1.csv')
finish_time = time.time()
print ''
print str(finish_time-start_time) + ' seconds elapsed in total.'
```
