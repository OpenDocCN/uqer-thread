# 数据可视化：几种图表的介绍

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np 
```

有时候，为了生成最优化的算法模型，必须对数据进行理解。最快、最有效的方式是通过数据的可视化来加强对数据的理解。本文将介绍通过matplotlib来可视化数据。本文使用的数据集是[第一个机器学习项目——鸢尾花](https://uqer.datayes.com/v3/community/share/5c22f4636105670121c8248f)使用的数据集。

首先介绍三种单一图表，分别是：直方图、密度图䄦箱线图。

## 1.1 直方图

直方图(Histogram)，又称质量分布图，是一种统计报告图，由一系列高度不等的纵向条纹或线段表示数据分布的情况。一般用横轴表示数据类型，纵轴表示分布情况。直方图可以很直观地展示每个属性的分布情况。它是一个连续变量（定量变量）的概率分布的估计，并且被卡尔·皮尔逊（Karl Pearson）首先引入。代码实现如下，我们可以看到这些属性的分布。

```python
filename = "iris.data.csv"
names = ["separ-length","separ-width","petal-length","petal-width","class"]
data = pd.read_csv(filename,names=names)
data.hist()
plt.show()
```

## 1.2 密度图

密度图表现与数据值对应的边界或域对象的一种理论图形表示方法。一般用于呈现连续变量，而如果变量是离散变量。密度图类似于直方图进行抽象，用平滑的线来描述数据的分布，这也是一种描述数据分布的图表。代码实现如下。

```python
filename = "iris.data.csv"
names = ["separ-length","separ-width","petal-length","petal-width","class"]
data = pd.read_csv(filename,names=names)
data.plot(kind = "density",subplots = True, layout = (2,2),sharex = False, sharey = False )
plt.show()
```

## 1.3 箱线图

箱形图（Box-plot）又称为盒须图、盒式图或箱线图，是一种用作显示一组数据分散情况资料的统计图。因形状如箱子而得名。在各种领域也经常被使用，常见于品质管理。它主要用于反映原始数据分布的特征，还可以进行多组数据分布特征的比较。箱线图的绘制方法是：先找出一组数据的最大值、最小值、中位数和两个四分位数；然后， 连接两个四分位数画出箱子；再将最大值和最小值与箱子相连接，中位数在箱子中间。代码实现如下。

```python
filename = "iris.data.csv"
names = ["separ-length","separ-width","petal-length","petal-width","class"]
data = pd.read_csv(filename,names=names)
data.plot(kind = "box",subplots = True, layout = (2,2),sharex = False, sharey = False )
plt.show()
```

接下来介绍两种图表，以显示不同属性之间的关联关系：相关矩阵图和散点矩阵图。

## 2.1 相关矩阵图

相关矩阵图主要是用来展示两个不同属性之间相互影响的程度。如果两个属性按照相同的方向变化，这说明是正向影响。如果两个属性朝相反的方向变化，这说明是反响影响。吧所以属性两两影响的关系展示出来的图表就叫相关矩阵图。矩阵图法就是从多维问题的事件中找出成对的因素，排列成矩阵图，然后根据矩阵图来分析问题，确定关键点。它是一种通过多因素综合思考来探索问题的好方法。代码实现如下。

```python
filename = "iris.data.csv"
names = ["separ-length","separ-width","petal-length","petal-width","class"]
data = pd.read_csv(filename,names=names)
correlations = data.corr()
print(correlations)
fig = plt.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(correlations, vmin = -1, vmax = 1)
fig.colorbar(cax)
ticks = np.arange(0,4,1)
ax.set_xticks(ticks)
ax.set_yticks(ticks)
ax.set_xticklabels(names)
ax.set_yticklabels(names)
plt.show()
```

## 2.2 散点矩阵图

散点矩阵图表示因变量随自变量变化的大致趋势，据此可以选择合适的函数对数据点进行拟合。散点矩阵图由两组数据构成多个坐标点，考察坐标点的分布，可以判别两个变量之间是否存在某种关联或 总结坐标点的分布模式。散点矩阵图将序列显示为一组点，值由点在图标中的位置表示，类别由图标中的不同标记表示。散点矩阵图通常用于比较跨类别的聚合数据。当同时考察多个变量的相关关系时，若一一绘制他们的简单散点图将十分麻烦。此时可利用散点矩阵图来绘制各个变量之间的散点图，这样就可以快速展现多个变量之间的主要相关性，这在进行多元线性回归时显得尤为重要。代码实现如下。

```python
filename = "iris.data.csv"
names = ["separ-length","separ-width","petal-length","petal-width","class"]
data = pd.read_csv(filename,names=names)
pd.scatter_matrix(data)
plt.show()
```
