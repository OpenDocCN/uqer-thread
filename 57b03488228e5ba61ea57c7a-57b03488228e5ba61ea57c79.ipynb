# 说明

一般而言，Category（激励计划-股权激励）的数量少于Category（进展公告-股权激励）。

主要是如果管理办法如果不修改的话，那么一个比如3年期的股权激励计划，就只会在Category（激励计划-股权激励）中公告一次，其他的会在Category（进展公告-股权激励）中公告。

要具体分析，大体就是把两个加起来，做一个并集。

```python
set1 = DataAPI.ReportByCategoryGet(beginDate=u"20090105",endDate=u"",Category=u"进展公告-股权激励",field=u"",pandas="1")
set2 = DataAPI.ReportByCategoryGet(beginDate=u"20090105",endDate=u"",Category=u"激励计划-股权激励",field=u"",pandas="1")
df = set1.append(set2)
```

```python
# 现在根据year分组，取每一组中的唯一的ticker
res = df.groupby(['year','ticker']).count()
res = res.reset_index().groupby('year')['ticker'].count()
# 上面这两句代码烂透了，应该有更方便的写法。
ax = res.plot.bar()
ax.set_title(u"采用股权激励历史数量",fontproperties = font)
```

# 2.具体不同方案的分析

```python
xz = df[df['title'].str.contains('限制性')]
res = xz.groupby(['year','ticker']).count()
res = res.reset_index().groupby('year')['ticker'].count()
# 上面这两句代码烂透了，应该有更方便的写法。
ax = res.plot.bar()
ax.set_title(u"限制性股票",fontproperties = font)
```

```python
qq = df[df['title'].str.contains('期权')]
res = qq.groupby(['year','ticker']).count()
res = res.reset_index().groupby('year')['ticker'].count()
# 上面这两句代码烂透了，应该有更方便的写法。
ax = res.plot.bar()
ax.set_title(u"期权",fontproperties = font)
```

```python
zz = df[df['title'].str.contains('增值权')]
res = zz.groupby(['year','ticker']).count()
res = res.reset_index().groupby('year')['ticker'].count()
# 上面这两句代码烂透了，应该有更方便的写法。
ax = res.plot.bar()
ax.set_title(u"增值权",fontproperties = font)
```

```python
# 看一眼为何还有那么多剩下的
used = pd.concat([xz,qq,zz])
df[~df.index.isin(used.index)]
# 大体就是，有些公告名没有写出到底发行的是什么方案
```
