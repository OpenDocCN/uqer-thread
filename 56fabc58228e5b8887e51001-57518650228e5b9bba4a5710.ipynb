# 程序员学量化投资（三）：区间内低买高卖

上篇文章学习了在指定的低价格点位买入股票，从而获取盈利。但是实际操作中我们如何确定什么是低价位，什么是高价位呢？
这个时候，我们就要用到优矿强大的历史数据功能，让计算机帮助我们分析行情，得到适当的低位价格和高位价格。
本策略通过统计30日周期内的波段高位和波段低位，得到价格区间。在区间内下方20%的区域买入操作，在上方20%的区域卖出，从而实现低买高卖的目的。

这段代码思想上应该是没问题的，遗憾的是优矿的回测程序有BUG。在调用account.get_daily_attribute_history 方法后，account.referencePrice 获取到的参考价变成了日线的收盘价而不是分钟线价格，导致价格过高无法买入，所以没有回测结果。
希望工作人员能尽快修复这个问题。
