# 机器学习示例

# Python Codes for Popular ML Algorithms
Below we provide sample Python codes, demonstrating use popular Machine Learning algorithms.

# Lasso

```python
from sklearn.linear_model import Lasso
model = Lasso(alpha=0.1)
model.fit([[-1,-1],[0,0],[1,1]], [-1,0,1])
```

```python
print(model.coef_)
```

```python
print(model.intercept_)
```

```python
model.predict([[3,-3]])
```

# Ridge

```python
from sklearn.linear_model import Ridge
model = Ridge(alpha=0.1)
model.fit([[-1,-1],[0,0],[1,1]], [-1,0,1])
```

```python
print(model.coef_)
```

```python
print(model.intercept_)
```

```python
model.predict([[3,-3]])
```

# ElasticNet弹性网络回归

```python
from sklearn.linear_model import ElasticNet
model = ElasticNet(alpha=0.1)
model.fit([[-1,-1],[0,0],[1,1]], [-1,0,1])
```

```python
print(model.coef_)
```

```python
print(model.intercept_)
```

```python
model.predict([[3,-3]])
```

# K-Nearest Neighbors

```python
from sklearn.neighbors import NearestNeighbors
import numpy as np
X = np.array([[-1, -2], [-2, -2], [-3, -5], [1, 1], [2, 2], [4, 4]])
model = NearestNeighbors(n_neighbors=2, algorithm='ball_tree').fit(X)
distances, indices = model.kneighbors([[0,0]])
distances
```

```python
indices
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(penalty='l2')
model.fit([[-2,-3], [1,0], [1,1]], [1, 0, 1])
```

```python
print(model.coef_)
```

```python
print(model.intercept_)
```

```python
model.predict([[3,3]])
```

# SVM

```python
from sklearn.svm import SVC
import numpy as np
X = np.array([[-3, -2], [-4, -5], [3, 4], [4, 5]])
y = np.array([1, 1, 2, 2])
model = SVC()
model.fit(X, y)
```

```python
print(model.predict([0,0]))
```

# Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
import numpy as np
X = np.array([[-3, -2], [-4, -5], [3, 4], [4, 5]])
y = np.array([1, 1, 2, 2])
model = RandomForestClassifier()
model.fit(X, y)
```

```python
print(model.predict([0,0]))
```

# K-Means

```python
from sklearn.cluster import KMeans
import numpy as np
model = KMeans(n_clusters=2, random_state=0).fit(X)
model.labels_
```

```python
model.predict([[1,2], [-1,-1]])
```

```python
model.cluster_centers_
```

# PCA

```python
from sklearn.decomposition import PCA
import numpy as np
model = PCA(n_components=2)
model.fit(X)
```

```python
print(model.explained_variance_ratio_)
```
