```python
import pandas as pd
import datetime as dt
```

```python
data = DataAPI.MktEqudGet(ticker=u"600680",beginDate=u"20140101",endDate=u"20160427",isOpen="1",field=u"tradeDate,closePrice",pandas="1")
data['tradeDate'] = data.apply(lambda row: (dt.datetime.strptime(row["tradeDate"], "%Y-%m-%d")), axis = 1) #因为DataAPI所有日期都是以字符串形式存储的，因此此处需要转换成日期时间型,这里很重要！
data = data.set_index('tradeDate')
data[:10]
```

```python
data['maShort'] = pd.rolling_mean(data.closePrice.shift(1), 5)
data['maLong']  = pd.rolling_mean(data.closePrice.shift(1), 20)
data[:25]
```

```python
# calculate the order signals
data['signal'] = 0 # default to 0
data.loc[data.maShort > data.maLong, 'signal'] = -1 #卖
data.loc[data.maShort < data.maLong, 'signal'] = 1  #买
data
```

```python
#Computing Returns
data['PctChg'] = data.closePrice.pct_change().shift(-1)
data
```

```python
# calculate the returns
data['ret'] = data.PctChg * data.signal
data
```

```python
#Cumulative returns and the result of the strategy
      
# calculate and report the cumulative returns
data['cumret'] = (1 + data.ret).cumprod() - 1
data
```

```python
# show graph
data['cumret'].plot(figsize=(12,4));
```
