# 1、Quartz3的回测流程及重要类的关系

<p><strong>Quartz3</strong> 基于面向对象的思想设计，有如下重要类定义。</p>
<ul>
<li>TradingStrategy<ul>
<li>策略对象，包含了策略算法函数 handle_data，被策略框架调度。</li>
</ul>
</li>
<li>SimulationParameter<ul>
<li>回测参数对象，包含了回测时间、频率、策略基准等回测基础参数，决定了策略框架的调度方式。</li>
</ul>
</li>
<li>AccountConfig<ul>
<li>账户配置对象，包含了账户配置信息，用于初始化 StockAccount 等交易账户对象。</li>
</ul>
</li>
<li>Context<ul>
<li>策略运行环境对象，包含了策略运行时的时间、数据等信息，由策略框架调度。</li>
</ul>
</li>
<li>StockAccount<ul>
<li>股票账户对象，包含了账户持仓、订单簿等信息，可以用来进行下单和查询操作，由 AccountConfig 配置。</li>
</ul>
</li>
<li>Order<ul>
<li>订单对象，由 StockAccount 持有，包含一次订单委托的信息。</li>
</ul>
</li>
<li>Position<ul>
<li>持仓对象，由 StockAccount 持有，包含账户的持仓信息。</li>
</ul>
</li>
</ul>

我个人认为回测程序的流程如下：
    1、实例化SimulationParameter类，传入自己设定的回测参数。实例化AccountConfig类，传入自己设定的账户参数。
    2、通过initialize函数实例化策略运行环境（Context），此时可以为Context添加新的属性，自定义各种变量。
    3、通过handle_data函数导入策略
    4、实例化TradingStrategy，通过initialize、handle_data分别传入的初始参数和策略，模拟交易、执行回测。

quartz是优矿编写的函数库(Libraty)也称做包(Package),在python中检测其类型为模块(Module),通过print函数打印quartz，可以看到其含有__init.pyc,故其是库

```python
print(quartz)
print(quartz.__class__)
```

1、TradingStrategy是quartz.context.strategy中的类，直接继承自Object类，其接收initialize, handle_data, post_trading_day三个函数参数传入

```python
class TradingStrategy(initialize, handle_data, post_trading_day)
```

```python
print(quartz.TradingStrategy)
print(quartz.TradingStrategy.__class__)
print(quartz.TradingStrategy.__bases__)
```

2、SimulationParameters是quartz.context.parameters中的类，直接继承自Object类，其接收start, end, benchmark, universe, freq,max_history_window, accounts 7个属性参数传入

```python
class SimulationParameters(start, end, benchmark='HS300', universe=[], freq='d', max_history_window=30, accounts={})
```

```python
print(quartz.SimulationParameters)
print(quartz.SimulationParameters.__class__)
print(quartz.SimulationParameters.__bases__)
```

3、AccountConfig属于quartz.account.account库中的类，继承自Object类

```python
class AccountConfig(account_type, capital_base, position_base, cost_base, commission, slippage)
```

```python
print(AccountConfig)
print(AccountConfig.__class__)  
print(AccountConfig.__bases__)
```
