# 2千多支股票怎么选，怎么不用 MACD 来择股择时

### MACD 金叉，死叉，二次金叉 -> 择股择时

- step 1. 获取股票池
- step 2. 股票池筛选：价格，上市日期【07年之后】, PE, PB, MV
- step 3. 计算某只股票目前是否出现金叉
- step 4. 其他辅助：
	+ 多空对比
    - 看看是否用顶背离，是否严重
	- 在5日线和10日线上逗留，周线，月线也可参考
    - 要比较灵活的股，量大，振幅大
    - 周线可以参考参考，造假程度更难
    - 最近新闻情况
- step 5. 投钱买啊。。。。。。 

### 优化记录

- MACD金叉滞后性，改变思维，不选MACD金叉的股票，选预计会出现MACD金叉的股票，而这里的预计方法，就是看MACD的趋势，虽然MACD为负，但只要MACD是呈正V行反转，并且已经反转到一个值，那就可以大胆预测即将出现MACD金叉，作为买点；
- MACD金叉位置，低位金叉一般比高位金叉有效
- MACD V型拐点的计算方式也有多样：
	+ 当前MACD值的绝对值小于此区间中前2个的绝对值；    
    + 当前MACD值处于此区间的底部，即当前MACD绝对值小于前一日MACD绝对值，且前一日MACD绝对值是此区间中MACD绝对值中的最大值
- MACD 大角度金叉：
	+ 即MACD金叉的角度很大，表示多空强劲
    + 未发生金叉，MACD趋向金叉动能强经

```python
# 所需包
import pandas as pd
import CAL.PyCAL as c
import seaborn
import matplotlib as mplt
from datetime import datetime as dt
import lib.factor as myfactor

# 变量定义
# 这里的today意义比较特殊，因为公司的因子库是每日上午5点更新，所以准备选股票的话，today就要设置为昨天。
# slice_step 的设置是因为DataAPI受限，一次不能取所有ticker的factor；
today = '20150518'
slice_step = 50
```

```python
### 1. step 1
def get_all_ticker():    
    tmp = DataAPI.EquGet(equTypeCD="A", listStatusCD="L", field=['ticker', "listDate", "ListSector"])
    equ_pool = DataAPI.MktEqudLatelyGet(field=["ticker", "exchangeCD", "secShortName","preClosePrice", "openPrice", "dealAmount", 
                                "highestPrice", "lowestPrice", "closePrice", "negMarketValue", "marketValue"])
    tmp = pd.merge(left=equ_pool, right=tmp, left_on="ticker", right_on="ticker")
    return tmp[tmp['ListSector'] != '创业板']

## 市值，市盈率，市净率
def get_factors(equ_pool, tradeDate):
    '''factors: PB, PE'''
    tmp = pd.DataFrame()
    tlist = list(equ_pool['ticker'])
    for i in [tlist[j:j+slice_step] for j in range(0, len(tlist), slice_step)]:
        tmp = tmp.append(DataAPI.MktStockFactorsOneDayGet(tradeDate=tradeDate, ticker=i, field = ['ticker','PB','PE']))
    equ_pool = pd.merge(left=equ_pool, right=tmp, left_on="ticker", right_on="ticker")
    return equ_pool

raw_pool = get_all_ticker()
raw_pool = get_factors(raw_pool, today)
```

```python
## step 2. 价格，PB,PE,MV,ListDate
def filter_price_date(equ_pool, low=2, high=30, PB=30, PE=200, MV=5*10**10, year=2007):
    low_bound = equ_pool[equ_pool["highestPrice"] != 0]
    low_bound = low_bound[equ_pool["closePrice"] > low]
    high_bound = low_bound[low_bound["closePrice"] < high]
    date_bound = high_bound[high_bound["listDate"] > str(year)]
    factor_bound = date_bound[date_bound['PB'] < PB]
    factor_bound = factor_bound[factor_bound['PE'] < PE]
    factor_bound = factor_bound[factor_bound['marketValue'] < MV]    
    return factor_bound
```

```python
## step 3.1 计算金叉，死叉
## 有关金叉，死叉的参数：
## 1. ema：short 和 long的取值，目前大多数软件都取12天和26天，那么目前的震荡牛市里，是不是需要把时间缩短一点呢？
## 2. dea：天数为9天，这个9的来源？应该是统计的吧，那统计值是否适用于当今市场，是否适用于所有股票呢？
## 3. macd：这里只是计算出了简单的交叉，可简单的交叉就能算金叉，死叉吗，应该可以从夹角这个方面去思考思考吧？
############
def cal_ema(equ_pool, n):
    '''计算N日EMA指数平均值，注意处理停盘情况'''
    ema = [equ_pool['closePrice'][0]] * len(equ_pool)
    for i in range(1, len(equ_pool)):
        if equ_pool['turnoverVol'][i] == 0:
            ema[i] = ema[i-1]
        else:
            ema[i] = ema[i-1] * (n-1)/(n+1) + equ_pool['closePrice'][i] * 2/(n+1)
    equ_pool['EMA_'+str(n)] = ema
    return equ_pool
############
def cal_diff(ticker, listDate, shortNumber=12, longNumber=26):
    '''计算DIFF偏移值'''
    tmp = DataAPI.MktEqudAdjGet(ticker=ticker, beginDate=listDate.replace('-',''), field=["ticker","closePrice","tradeDate","turnoverVol"])
    tmp = cal_ema(tmp, shortNumber)
    tmp = cal_ema(tmp, longNumber)
    tmp['DIFF'] = tmp["EMA_" + str(shortNumber)] - tmp["EMA_" + str(longNumber)]
    return tmp
```

```python
def cal_dea(equ_pool, n=9):
    '''计算DEA差离平均值'''
    dea = [equ_pool['DIFF'][0]] * len(equ_pool)
    for i in range(1, len(equ_pool)):
        if equ_pool['turnoverVol'][i] == 0:
            dea[i] = dea[i-1]
        else:
            dea[i] = dea[i-1] * (n-1)/(n+1) + equ_pool['DIFF'][i] * 2/(n+1)
    equ_pool['DEA'] = dea
    return equ_pool
############
def cal_bs(equ_pool):
    '''根据MACD计算买卖点'''
    macd = equ_pool['MACD']
    bs = [0] * len(macd)
    try:
        last = macd[1]
    except:
#         print equ_pool
        pass
    for i in range(2, len(macd)):
        current = macd[i]
        if last * current < 0:
            bs[i] = 1 if current > 0 else 2
        last = current
    equ_pool['BuySell'] = bs
    return equ_pool
############
def cal_macd(ticker, listDate):
    '''计算MACD指数平滑平均值'''
    tmp = cal_diff(ticker, listDate)
    macd_data = cal_dea(tmp)
    macd_data['MACD'] = 2 * (macd_data['DIFF'] - macd_data['DEA'])
    return macd_data
```

```python
def is_vtrans_v1(ticker):
    '''
    cal_bs_vtrans 的临时解决方案，就看ticker是否满足情况：
    1. 当前处于即将迎来下一个金叉的区域
    2. 当前MACD值的绝对值小于当前区域的前2个MACD值的绝对值【这点可以预示最近将来2天，有可能出现MACD金叉】
    传入的ticker是一个包含MACD，上市日期等列数据的一个DataFrame；
    若满足，则返回True，否则返回False
    '''
    macd = list(ticker['MACD'])
    # 若当前MACD是正值，那下一次是会出现死叉，不是金叉
    if macd[-1] > 0:
        return False
    # 当前MACD是负值，下一次会出现金叉
    else:
        # 找到此次死叉的起点
        start = -1
        while macd[start] < 0:
            start -= 1
        start += 1
        # 获取此次死叉->金叉区间
        section = macd[start:]
        # 先简单判断，看最近一次的macd值是否大于此次区间开始后的第一个macd值，若如此，说明此次macd死叉区间即将走到尽头，金叉即将出现
        # 这里还有其他很多优化算法，不过现在先暂时不用，由简入深，一步一步来，不能没有前戏就直接高潮啊
        if section[-1] > section[0]:
            return True
        else:
            return False
########
def is_vtrans_v2(ticker):
    '''
    临时解决方案，就看ticker是否满足情况：
    1. 当前MACD值处于此区间的底部，即当前MACD绝对值小于前一日MACD绝对值，且前一日MACD绝对值是此区间中MACD绝对值中的最大值
    传入的ticker是一个包含MACD，上市日期等列数据的一个DataFrame；
    若满足，则返回True，否则返回False
    '''
    macd = list(ticker['MACD'])
    # 若当前MACD是正值，那下一次是会出现死叉，不是金叉
    if macd[-1] > 0:
        return False
    # 当前MACD是负值，下一次会出现金叉
    else:
        # 找到此次死叉的起点
        start = -1
        while macd[start] < 0:
            start -= 1
        start += 1
        # 获取此次死叉->金叉区间
        section = macd[start:]
        # 检查当天MACD是否大于前一日MACD【负值】
        # 检查前一日MACD是否是此区间的最小值【负值】
        if len(section)>2 and section[-1] > section[-2] and section[-2] == min(section):
            return True
        else:
            return False
```

```python
## step 3.2 我勒个去，最近2天哪些股票有金叉啊。。。。
def get_potential(equ_pool, nshort=2, nlong=7):
    '''计算最近n天哪些股票出现了金叉，返回一次金叉，二次金叉，低位金叉'''
    one_cross = []
    two_cross = []
    low_cross = []
    tickers = list(equ_pool['ticker'])
    listDate = list(equ_pool['listDate'])
    length = len(tickers)
    for i in range(length):
        ticker = tickers[i]
        tmp = cal_macd(ticker, listDate[i])
        tmp = cal_bs(tmp)
        true_false_short = list(tmp['BuySell'][-nshort:])
        true_false_long = list(tmp['BuySell'][-nlong:])
        diff = list(tmp['DIFF'])
        dea = list(tmp['DEA'])
        ####
        if True in true_false_long:  # 存在金叉
            if True in true_false_short:   # 一次金叉
                one_cross.append(ticker)
            if True in true_false_short and (diff[-1] < 0.2 or dea[-1] < 0.2):   # 低位金叉
                low_cross.append(ticker)
            if true_false_long.count(True) > 1 and true_false_long.count(False) > 0:    # 二次金叉
                two_cross.append(ticker)
    print 'one_cross: {}\n    {}, \nlow_cross: {}\n    {}, \ntwo_cross: {}\n    {}'.format(
        len(one_cross), one_cross, len(low_cross), low_cross, len(two_cross), two_cross)
    one_cross = one_cross if one_cross else [None]
    low_cross = low_cross if low_cross else [None]
    two_cross = two_cross if two_cross else [None]
    return (pd.merge(left=equ_pool, right=pd.DataFrame(data=one_cross, columns=['ticker']), left_on="ticker", right_on="ticker"),
         pd.merge(left=equ_pool, right=pd.DataFrame(data=low_cross, columns=['ticker']), left_on="ticker", right_on="ticker"),
         pd.merge(left=equ_pool, right=pd.DataFrame(data=two_cross, columns=['ticker']), left_on="ticker", right_on="ticker"))
############
def macd_vtrans(equ_pool):
    '''
    条件 1. 最近一次出现的是MACD死叉
    条件 2. MACD值出现反转，反转的算法单独实现
    返回：返回过滤后的equ_pool，里面的每支ticker都满足条件1，2
    '''
    vtrans_v1 = []
    vtrans_v2 = []
    tickers = list(equ_pool['ticker'])
    listDate = list(equ_pool['listDate'])
    length = len(tickers)
    
    for i in range(length):
        ticker = tickers[i]
        # 计算每支股票MACD值
        tmp = cal_macd(ticker, listDate[i])
        # 如果最近几天出现了MACD V型拐点指示的买点，则选择这只股票
        if is_vtrans_v1(tmp):
            vtrans_v1.append(ticker)
        if is_vtrans_v2(tmp):
            vtrans_v2.append(ticker)
    print 'vtrans_v1: {}\n    {}, \nvtrans_v2: {}\n    {}, \n'.format(len(vtrans_v1), vtrans_v1, len(vtrans_v2), vtrans_v2)
    vtrans_v1 = vtrans_v1 if vtrans_v1 else [None]
    vtrans_v2 = vtrans_v2 if vtrans_v2 else [None]
    return (pd.merge(left=equ_pool, right=pd.DataFrame(data=vtrans_v1, columns=['ticker']), left_on="ticker", right_on="ticker"),
         pd.merge(left=equ_pool, right=pd.DataFrame(data=vtrans_v2, columns=['ticker']), left_on="ticker", right_on="ticker"))
## 打印相关信息，行业，主题，新闻
def print_info(equ_pool):
    tmp = myfactor.build_themes(equ_pool)
    return myfactor.filter_news(equ_pool, 5)
```
