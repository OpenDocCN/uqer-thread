# Calendar

```python
import pandas as pd
class TradeCalendar():
    def __init__(self, beginDate="20100101", endDate="20221231"):
        self.beginDate = beginDate
        self.endDate = endDate
        self.load_calendar()
        self.tradeDays = self.calendar_date.keys()
        
    def load_calendar(self):
        """初始化日期"""
        self.calendar_date = {}
        self.calendar_index = DataAPI.TradeCalGet(exchangeCD=u"XSHG",beginDate=self.beginDate,endDate=self.endDate,\
                                       isOpen=u"1",field=u"calendarDate",pandas="1")['calendarDate']
        for index, date in enumerate(self.calendar_index):
            self.calendar_date[date] = index
    
    def is_trade_day(self, date):
        """判断某一个日期是否为交易日"""
        if type(date) == str:
            return date in self.tradeDays
        return date.strftime("%Y-%m-%d") in self.tradeDays
    
    def gap_between_days(self, day1, day2):
        """判断两个交易日间隔了几个交易日"""
        assert (self.is_trade_day(day1) and self.is_trade_day(day2)), "Invalid Trade Day"
        if not (type(day1) == str and type(day2) == str):
            return self.calendar_date[self.str_time(day2)] - self.calendar_date[self.str_time(day1)]
        return self.calendar_date[day2] - self.calendar_date[day1]
        
    
    def get_n_days_before(self, today, N):
        """得到前第N个交易日，不包括今日（可理解为开盘前看）,N>0看之前，N<0看未来"""
        assert self.is_trade_day(today), "Invalid Trade Day"
        if type(today) != str:
            today = self.str_time(today)
        return self.calendar_index[self.calendar_date[today]-N]
    
    def get_prev_n_tradeDays(self, today, N):
        """得到前面连续N个交易日，不包括今日,N>0看之前，N<0看未来"""
        assert self.is_trade_day(today), "Invalid Trade Day"
        if type(today) != str:
            today = self.str_time(today)
        cur_index = self.calendar_date[today]
        return self.calendar_index[min(cur_index-N, cur_index): max(cur_index, cur_index-N)]
    
    def dt(self, string):
        """把一个日期的string转换成datetime"""
        return pd.to_datetime(string)

    def str_time(self, date):
        """Timestamp转成时间"""
        return date.strftime("%Y-%m-%d")
```
