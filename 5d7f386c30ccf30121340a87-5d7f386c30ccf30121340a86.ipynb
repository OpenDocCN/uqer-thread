# 根据股票名称查找股票

```python
def findStock(s):
    df=DataAPI.EquGet(secID=u"",ticker=u"",equTypeCD=u"A",listStatusCD=u"",field=u"",pandas="1")[['secID','secShortName']]
    l=[]
    for index,row in df.iterrows():
        if s in row['secShortName']:
            l.append(row['secID'])
    ldf=DataAPI.EquGet(secID=l,ticker=u"",equTypeCD=u"A",listStatusCD=u"",field=u"",pandas="1")
    return ldf

findStock('石油')
```
