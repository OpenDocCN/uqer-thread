# 置之死地而后生 - 大跌反弹策略的验证

第一次写策略，写不了太复杂的，如果有不对的地方，欢迎大家拍砖，谢谢！

本策略基于如下思想：

1 股票走势总是上下波动的，不会一直上涨或者下跌

2 在一次大跌之后，或多或少会有反弹

3 在连续两天累计跌幅超过7%的时候买入，然后当股价涨回买入价的3%马上卖出。

具体策略逻辑如下：

* 监控股票池为沪深300.
 
* benchmark是沪深300指数。

* 对符合买入条件的股票平均买入。

* 建仓：如果今天的股价比两天前的股价低超过了7%，满仓。

* 减仓：如果股价比买入成本价高出3%，清仓。

```python

# coding: utf-8

# In[ ]:

import numpy as np
import quartz
from quartz.api import *

import pandas as pd
import numpy as np
from datetime import datetime
from matplotlib import pylab


# 开始
start = datetime(2006, 1, 1)
end   = datetime(2016, 3, 29)
benchmark = 'HS300'							# 策略参考标准
universe = set_universe('HS300')
#universe = ['601088.XSHG']
capital_base = 1000000
refresh_rate = 1


def initialize(account):
    pass
      
def handle_data(account):
    signals = []
    
    hist = account.get_attribute_history('closePrice',5)
    for stock in account.universe:
        
        position = account.valid_secpos.get(stock, 0)
        price = hist[stock][-1]
        
        if stock in account.valid_seccost.keys():
            if (price / account.valid_seccost[stock]) > 1.03 and position != 0:
#                print ("selling price = %f"%price)
                order_to(stock, 0)
                continue
            continue
        if (price / hist[stock][-3]) < 0.93:
            # 满足买入条件，加入signals列表中
            signals.append(stock)
#            print ("buying price = %f"%price)
            
    for stock in signals:
        # 平均买入signals列表中的股票
        amount = account.cash/len(signals)/account.referencePrice[stock]
        order(stock, amount)
        
    observe("hold_rate",1.0 - account.cash/account.referencePortfolioValue)    
    
bt, acct = quartz.backtest(start = start,
             end = end,
             benchmark = benchmark,
             universe = universe,
             capital_base = capital_base,
             initialize = initialize,
             handle_data = handle_data,
             refresh_rate = refresh_rate)
perf = quartz.perf_parse(bt, acct)

out_keys = ['annualized_return', 'volatility', 'information_ratio', 'sharpe', 'max_drawdown', 'alpha', 'beta']

for k in out_keys:
    print '%s:%s' % (k, perf[k])
    
print bt.keys()

bt.index = bt["tradeDate"]

bt.plot(y=[u'hold_rate'], figsize=(14, 7), linewidth='1')


perf['cumulative_returns'].plot()
perf['benchmark_cumulative_returns'].plot()
pylab.legend(['hold_rate','current_strategy', 'benchmark'])
```

回测了自2006年1月1日至2016年3月29日全部的数据，可以看出该策略在不同的时期表现不一样。2009年之前该策略的表现和大盘的表现十分贴近，并能够微微战胜大盘。2009年到2014年末的大盘横盘震荡时期，该策略表现还不错，只回测这一段时间，能跑赢大盘。表现最好的时候是2015年的那波大涨，大大超过了大盘的涨幅，但是在随后的大跌中也承受力严重的回撤。
