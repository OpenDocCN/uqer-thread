```python
import numpy as np
import pandas as pd 
import statsmodels.api as sm
from scipy import stats
```

```python
cal = DataAPI.TradeCalGet(exchangeCD=u"XSHG",beginDate=u"",endDate=u"",field=u"",pandas="1")
cal = cal[cal['isMonthEnd'] == 1]
pre_cal = cal.shift(1)
```

```python
def get_month_factor(date):
    ## 当前股票
    tickers_a = set_universe('A',date= date)
    # 获取因子数据-排序-分组
    
    ## 取规模因子-市值的大小
    ### size 分组
    ME = DataAPI.MktEqudGet(secID=tickers_a ,ticker=u"",tradeDate= date ,isOpen=u"",field=u"ticker,marketValue" ,pandas="1").dropna()
    ME50=np.percentile(ME['marketValue'],50)                                     
    S = ME[ME['marketValue']<=ME50]['ticker'].tolist()                           
    B = ME[ME['marketValue']>ME50]['ticker'].tolist()
    size_dict = {k:'s' for k in S}
    size_dict.update({k:'b' for k in B})

    ## 取估值因子-账面市值比
    BP = DataAPI.MktStockFactorsOneDayGet(tradeDate=date, secID=tickers_a, ticker=u'', field=u"ticker,PB",pandas="1").dropna()
    BP[['PB']] = 1/BP[['PB']]
    BP30=np.percentile(BP['PB'],30)
    BP70=np.percentile(BP['PB'],70)
    L = BP[BP['PB']<=BP30]['ticker'].tolist()
    H = BP[BP['PB']>BP70]['ticker'].tolist()
    M = list(set(BP['ticker'].tolist()).difference(set(L+H)))
    bp_dict = {k:'l' for k in L}
    bp_dict.update({k:'h' for k in H})
    bp_dict.update({k:'m' for k in M})

    ## 盈利因子-ROE 净利润 / 股东权益 
    PR = DataAPI.MktStockFactorsOneDayGet(tradeDate=date ,secID=tickers_a ,ticker=u'', field=u"ticker,ROE", pandas='1').dropna() 
    PR30 = np.percentile(PR['ROE'],30) 
    PR70 = np.percentile(PR['ROE'],70)
    LR = PR[PR['ROE']<=PR30]['ticker'].tolist()                               
    HR = PR[PR['ROE']>PR70]['ticker'].tolist()
    MR = list(set(PR['ticker'].tolist()).difference(set(LR + HR)))
    roe_dict = {k:'l' for k in LR}
    roe_dict.update({k:'h' for k in HR})
    roe_dict.update({k:'m' for k in MR})

    ## 投资因子- 资产增长率
    CA = DataAPI.MktStockFactorsOneDayGet(tradeDate=date, secID=tickers_a , ticker=u'', field=u"ticker,TotalAssetGrowRate", pandas='1').dropna() 
    CA30 = np.percentile(CA['TotalAssetGrowRate'],30) 
    CA70 = np.percentile(CA['TotalAssetGrowRate'],70)
    LC = CA[CA['TotalAssetGrowRate']<=CA30]['ticker'].tolist()                                
    HC = CA[CA['TotalAssetGrowRate']>CA70]['ticker'].tolist()
    MC = list(set(CA['ticker'].tolist()).difference(set(LC + HC)))
    tar_dict = {k:'l' for k in LC}
    tar_dict.update({k:'h' for k in HC})
    tar_dict.update({k:'m' for k in MC})

    # 计算因子值
    ## 计算月度收益
    date_index = cal[cal['calendarDate'] == date].index
    pre_date = pre_cal.loc[date_index]
    ticker_return = DataAPI.MktEqumGet(secID=tickers_a,ticker=u"",monthEndDate=u"",beginDate=date ,endDate=date ,isOpen=u"",field=u"ticker,return",pandas="1")
    
    ## 因子映射
    ticker_return['size'] = ticker_return['ticker'].map(size_dict)
    ticker_return['hml'] = ticker_return['ticker'].map(bp_dict)
    ticker_return['rmw'] = ticker_return['ticker'].map(roe_dict)
    ticker_return['cma'] = ticker_return['ticker'].map(tar_dict)
    # 按照流通股本加权
    ME_F = DataAPI.MktEqudGet(secID=tickers_a ,ticker=u"",tradeDate= date ,isOpen=u"",field=u"ticker,negMarketValue" ,pandas="1").dropna()
    ticker_return = pd.merge(ticker_return, ME_F, on= 'ticker').dropna()
    
    ## 分组加权数据
    smb_bp_df = ticker_return.groupby(['size','hml']).apply(lambda x: x['return'] * x['negMarketValue'] / x['negMarketValue'].sum() ).unstack().sum(axis = 1).unstack()
    smb_op_df = ticker_return.groupby(['size','rmw']).apply(lambda x: x['return'] * x['negMarketValue'] / x['negMarketValue'].sum() ).unstack().sum(axis = 1).unstack()
    smb_inv_df = ticker_return.groupby(['size','cma']).apply(lambda x: x['return'] * x['negMarketValue'] / x['negMarketValue'].sum() ).unstack().sum(axis = 1).unstack()
    
    ## 计算第一个因子 SMB
    smb_bp = smb_bp_df.loc['s'].mean() - smb_bp_df.loc['b'].mean()
    smb_op = smb_op_df.loc['s'].mean() - smb_op_df.loc['b'].mean()
    smb_inv = smb_inv_df.loc['s'].mean() - smb_inv_df.loc['b'].mean()
    smb = (smb_bp + smb_op + smb_inv) / 3
    ## 计算 HML
    hml = smb_bp_df.loc[:, 'h'].mean() - smb_bp_df.loc[:, 'l'].mean()
    ## 计算 RMW
    rmw = smb_op_df.loc[:, 'h'].mean() - smb_op_df.loc[:, 'l'].mean()
    ## 计算 CMA
    cma = smb_inv_df.loc[:, 'l'].mean() - smb_inv_df.loc[:, 'h'].mean()
    
    ### 计算市场收益
    rm = DataAPI.MktIdxmGet(beginDate=date, endDate=date ,indexID=u"000001.ZICN",ticker=u"",field=u"yReturn",pandas="1")['yReturn'].values[0] ## 无风险月收益
    return np.array([rm, smb, hml, rmw, cma])
```

```python
date = '2017-07-31'
factor_name = 'PB'
```

```python
## 25宫格
def month_group_return(date, factor_name = 'PB'):
    ## 当前股票
    tickers_a = set_universe('A',date= date)
    # 获取因子数据-排序-分组

    ## 取规模因子-市值的大小
    ### size 分组
    ME = DataAPI.MktEqudGet(secID=tickers_a ,ticker=u"",tradeDate= date ,isOpen=u"",field=u"ticker,marketValue" ,pandas="1").dropna()
    ME['qcut'] = pd.qcut(ME['marketValue'], 5, labels= [1, 2, 3, 4, 5])
    me_dict = {k:v for k,v in zip(ME['ticker'].values , ME['qcut'].values)}

    ### factor 分组                               
    To_factor = DataAPI.MktStockFactorsOneDayGet(tradeDate=date, secID=tickers_a, ticker=u'', field=u"ticker,%s"%factor_name ,pandas="1").dropna()
    To_factor['qcut'] = pd.qcut(To_factor[factor_name], 5, labels= [1, 2, 3, 4, 5])
    bp_dict = {k:v for k,v in zip(To_factor['ticker'].values, To_factor['qcut'].values)}

    ticker_return = DataAPI.MktEqumGet(secID=tickers_a,ticker=u"",monthEndDate=u"",beginDate=date ,endDate=date ,isOpen=u"",field=u"ticker,return",pandas="1")

    ticker_return['size'] = ticker_return['ticker'].map(me_dict)
    ticker_return[factor_name] = ticker_return['ticker'].map(bp_dict)
    # 按照流通股本加权
    ME_F = DataAPI.MktEqudGet(secID=tickers_a ,ticker=u"",tradeDate= date ,isOpen=u"",field=u"ticker,negMarketValue" ,pandas="1").dropna()
    ticker_return = pd.merge(ticker_return, ME_F, on= 'ticker').dropna()

    ticker_group = ticker_return.groupby(['size',factor_name]).apply(lambda x: x['return'] * x['negMarketValue'] / x['negMarketValue'].sum() ).unstack().sum(axis = 1)
    return ticker_group
```

```python
## 股改之后的因子值
factor_dict = {}
flag = 0
for i in cal['calendarDate']:
    if i > '2007-06' and i < '2015-08':
        factor_dict[i] = get_month_factor(i)
        if flag % 10 == 0:
            print('%s has done'%i)
        flag += 1
```

```python
## 从股改时间开始
factor_df = pd.DataFrame(factor_dict).T
factor_df.columns = ['rm', 'smb', 'hml', 'rmw', 'cma']
```

```python
def factor_regress(y, col = factor_df.columns):
    x = factor_df[col]
    x = sm.add_constant(x)
    model = sm.OLS(y, x)
    results = model.fit()
    c, p = results.params['const'] ,results.pvalues['const']
    return c, p
```

## Alpha检验与对比

```python
def A_alpha(df_c, df_r):
    alpha_1 = df_c.apply(lambda x: np.abs(x)).mean()
    alpha_2 = df_c.apply(lambda x: np.abs(x)).mean() / df_r.apply(lambda x: np.abs(x)).mean()
    alpha_3 = df_c.apply(lambda x: np.square(x)).mean() / df_r.apply(lambda x: np.square(x)).mean()
    return alpha_1, alpha_2, alpha_3
```
