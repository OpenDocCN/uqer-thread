·公式
$$FCFF = \frac{FCF+DELT}{EV_{latest}}$$
$$FCF = nocf_{ttm} - ociro_{ttm} + oocp_{ttm} - (fioaac_{ttm} + ncfsc_{ttm} - fioadc_{ttm} - ncdsc_{ttm})$$
$$DELT = fe_{ttm}(1 - itc_{ttm}/tp_{ttm})$$
·释义
nocf:N_CF_OPERATE_A(经营活动产生的现金流量净额)剔除缺失值
fioaac:PUR_FIX_ASSETS_OTH(构建固定资产/无形资产和其他长期资产支付的现金)缺失填0
fioadc:DISP_FIX_ASSETS_OTH(处置固定资产/无形资产和其他长期资产收回的现金净额)缺失填0
ociro：C_FR_OTH_OPERATE_A（收到其他与经营活动有关的现金）缺失填0
oocp:C_PATD_FOR_OTH_OP_A（支付其他与经营活动有关的现金）缺失填0
ncfsc:N_C_PAID_ACQUIS（取得子公司及其他营业单位支付的现金净额）缺失填0
ncdsc:N_DISP_SUBS_OTH_BIZ_C（处置子公司及其他营业单位收到的现金净额）缺失填0
itc:INCOME_TAX（所得税费用）缺失填0
tp:T_PROFIT（利润总额（亏损总额以"-"号填列））剔除缺失值和0
fe:FINAN_EXP（财务费用）缺失填0
FCF DELT 缺失填0

earngs
·名称
earning growth surprise,未预期盈利
·公式
$$sig = \frac{em_i - mean(em_{i-1},...,em_{i-8})}{std(em_{i-1},...,em_{i-8})}$$
·释义
$em_i$:当年(T年)当前季度净利润相对于T-1年当前季度净利润的同比增长率，如果过去8个季度经营活动数据少于6期，则对应股票的信号值为空。
增长率 = （T年q季度净利润 - T-1年q季度净利润）/abs(T-1年q季度净利润)
![图片注释](http://storage-uqer.datayes.com/60e2ba0714f512016c7829d7/ed93a618-fb4a-11eb-a5fb-0242ac140002)

gptgs
·名称
gross profit growth surprise,未预期毛利
·公式
$$sig = \frac{gi_i - mean(gi_{i-1},...,gi_{i-8})}{std(gi_{i-1},...,gi_{i-8})}$$

这里是营业利润，而非总营业利润（营业总收入减营业总成本）
增长率 = （T年q季度营业利润 - T-1年q季度营业利润）/abs(T-1年q季度营业利润)

若公司在T日发布盈余公告，则将T日收盘-T+1日最低价之间的跳空收益。
盈余公告包括：业绩报告，业绩快报，定期财报。
$$MktPEAD\\\_JOR=r_{c2l,t+1}-idx\\\_r_{c2l,t+1}$$
其中$r_{c2l,t+1}$,$idx\\\_r_{c2l,t+1}$为个股和指数T+1日最低价相对T日收盘价的收益率，指数为中证全指。
![图片注释](http://storage-uqer.datayes.com/60e2ba0714f512016c7829d7/71dcd5be-fb48-11eb-805f-0242ac140002)

若公司在T日发布盈余公告，则将T-2日收盘-T+1日收盘期间的累计超额收益。
盈余公告包括：业绩报告，业绩快报，定期财报。
$$MktPEAD\\\_ABR3=\prod_{i=T-1,T,T+1}^{}{(1+r_i)-1}$$
其中, $r_i$为超额收益，基准为中证全指。
特别说明，T日的因子为
$$MktPEAD\\\_ABR3=\prod_{i=T-1,T}^{}{(1+r_i)-1}$$
