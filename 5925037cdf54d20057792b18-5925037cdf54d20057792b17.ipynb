# Python数据分析与展示之matplotlib库2：基本绘图函数

绘图函数有很多，这里是几个简单的例子：饼图、直方图、极坐标图、散点图

```python
#绘制饼图
import matplotlib.pyplot as plt
labels = 'Frogs','Hogs','Dogs','Logs'
sizes  = [15,30,45,10]
explode=(0,0.1,0,0)

plt.pie(sizes,explode=explode,labels=labels,autopct='%1.1f%%',shadow=False,startangle=90)
plt.axis('equal')  #使饼图显示为圆形
plt.show()
```

```python
#绘制直图
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)
mu,sigma=100,20
a=np.random.normal(mu,sigma,size=100)

plt.subplot(311)
plt.hist(a,20,normed=1,histtype='stepfilled',facecolor='b',alpha=0.75)
plt.title('Histogram')
plt.show()

#调整第2个参数bin
plt.subplot(312)
plt.hist(a,10,normed=1,histtype='stepfilled',facecolor='b',alpha=0.75)
plt.title('Histogram')
plt.show()

plt.subplot(313)
plt.hist(a,40,normed=1,histtype='stepfilled',facecolor='b',alpha=0.75)
plt.title('Histogram')
plt.show()
```

```python
#绘制极坐标图
import numpy as np
import matplotlib.pyplot as plt 

N=20
theta=np.linspace(0.0,2*np.pi,N,endpoint=False)
radii=10*np.random.rand(N)
width=np.pi/4*np.random.rand(N)

ax=plt.subplot(111,projection='polar')
bars=ax.bar(theta,radii,width=width,bottom=0)

for r,bar in zip(radii,bars):
    bar.set_facecolor(plt.cm.viridis(r/10.))
    bar.set_alpha(0.5)
    
plt.show()
```

```python
#绘制散点图,面向对象的方法
import numpy as np
import matplotlib.pyplot as plt 

fig,ax=plt.subplots()
ax.plot(10*np.random.randn(100),10*np.random.randn(100),'o')
ax.set_title('simple scatter')
plt.show()
```
