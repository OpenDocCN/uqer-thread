# 个股和指数的回归 复杂版

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
from sklearn.linear_model import LinearRegression

a=DataAPI.EquRetudGet(listStatusCD=u"",secID=u"",ticker=u"000858",beginDate=u"20080101",dailyReturnNoReinvLower=u"",dailyReturnNoReinvUpper=u"",dailyReturnReinvLower=u"",dailyReturnReinvUpper=u"",endDate=u"20171231",isChgPctl=u"",field=u"tradeDate,secShortName,dailyReturnReinv",pandas="1")  #dailyReturnReinv就是每天涨跌幅

b=DataAPI.MktIdxdGet(indexID=u"",ticker=u"000001",tradeDate=u"",beginDate=u"20080101",endDate=u"20171231",exchangeCD=u"XSHE,XSHG",field=u"tradeDate,secShortName,CHGPct",pandas="1")
print a
print b

x=np.array(a[['dailyReturnReinv']])       # X：股票日收益率作为二维数组  
y=np.array(b['CHGPct'])                   # Y：指数日收益率作为一维数组

where_are_nan = np.isnan(x)               # 下面将数据中的空值NaN替换成 0
where_are_inf = np.isinf(x)
x[where_are_nan] = 0
x[where_are_inf] = 0

where_are_nan = np.isnan(y)
where_are_inf = np.isinf(y)
y[where_are_nan] = 0
y[where_are_inf] = 0                      

print 'X:'
print x                                         # 完成x,y数据的处理，打印出数据
print 'Y:'
print y                                   

clf=LinearRegression()                    
clf.fit(x,y)                                    # 将x,y代入线性回归模型，下面求出模型的关键参数

print '斜率：'
print clf.coef_
print '截距:'
print clf.intercept_  
print '判定系数R方:'
print clf.score(x,y)     
print '输入自变量(0.006)预测因变量:'
print clf.predict(0.006)    

plt.scatter(x,y,6,color='blue',marker='o',linewidth=0.1,alpha=0.8)
plt.xlabel('000858')
plt.ylabel('000001')
plt.axis([-0.1 ,0.1, -0.1, 0.1])                # 做出x,y的散点图

k=clf.coef_                                     # 下面画拟合直线
b=clf.intercept_
x1=np.linspace(-0.1,0.1,2500)
y1=k*x1+b
plt.plot(x1,y1,color='red',linewidth=5) 
plt.show()

```
