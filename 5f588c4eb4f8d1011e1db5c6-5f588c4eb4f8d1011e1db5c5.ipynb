# 基于IC加权和RSI过滤的财务选股

本策略基于“股权回报率”和“净利润增长率”两因子,通过IC加权构造复合因子，并过滤超涨（RSI>80）的股票，进行月度调仓。结果显示，基于该复合因子的选股策略，有明显的超额收益。

```python
import pandas as pd
import numpy as np
import datetime as dt
import matplotlib.pyplot as plt
# CAL.PyCAL中包含font
from CAL.PyCAL import *    
#绘图正常显示负号
plt.rcParams['axes.unicode_minus'] = False
#绘图正常显示中文
plt.rcParams['font.sans-serif'] = ['Microsoft YaHei']
```

以沪深300为股票池，时间范围为2019年初至2020年8月初。每月第一个交易日（调仓日）开盘时先卖出所有持仓股票，再按照前一交易日（参考日）的复合因子值从高到低排序，选出排名前20的股票，在开盘时等仓位买入，持有至下月第一个交易日开盘卖出。为了便于计算，本文剔除了在调仓期停牌的股票，即买入和卖出股票当日都不会停牌，相当于在买入时知道了哪些股票在未来卖出时会停牌，这是引入未来函数。由于停牌的发生频率较低，导致停牌的原因可能是利好、利空或中性的，且单只股票初始仓位只有5%，所以对评估策略表现影响不大。在实际操作中，若需要买入的股票停牌，则将该股票移除买入名单，等仓位买入非停牌股票。若需要卖出的股票在调仓期停牌，建议持有至下一调仓期卖出，相当于在本调仓期替换了一只股票。

```python
#定义所用数据最大时间范围
start_date = '2018-09-01'
end_date = '2020-09-01'
#定义股票池
universe = DynamicUniverse('HS300')
#用到的因子
factors = ['ROE','NetProfitGrowRate']
#投资组合的股票数量
n = 20
dates_df = DataAPI.TradeCalGet(exchangeCD="XSHG",beginDate=start_date,endDate=end_date,isOpen="1")
#设置参考因子值日期为每月最后一个交易日
reference_dates = dates_df[dates_df.isMonthEnd==1]['calendarDate'].tolist()
#设置调仓时期为每月第一个交易日
trade_dates = dates_df.loc[dates_df[dates_df.isMonthEnd==1].index+1,'calendarDate'].tolist()
```

财务指标好的股票通常会被资金追逐而上涨，但超涨的股票涨势难以维系，在建仓时应避免此类股票，故以RSI作为研判超涨的指标，过滤超涨股票。

```python
'''
定义函数：过滤不符合要求的股票
'''
def filter_stocks(date,stocks):
    #RSI不超过80
    rsi_df = DataAPI.MktStockFactorsOneDayGet(secID=stocks,tradeDate=date,field=['secID','RSI'])
    return rsi_df[rsi_df['RSI']<80]['secID'].tolist()
```

在构造复合因子时，需要对两财务因子加权求和，本策略采用滚动3期IC均值作为因子权重。

```python
'''
定义函数：计算因子IC
'''
def calculate_ic(start_date,end_date,factor_Series):
    #获取因子序列的股票名称
    stocks = factor_Series.index.tolist()
    #获取股票池期初开盘价
    start_prices_df = DataAPI.MktEqudAdjGet(secID=stocks,tradeDate=start_date,field=['secID','openPrice']).set_index('secID')
    #获取股票池期末开盘价
    end_prices_df = DataAPI.MktEqudAdjGet(secID=stocks,tradeDate=end_date,field=['secID','openPrice']).set_index('secID')
    #计算调仓期股票收益率
    revenue_Series = end_prices_df['openPrice']/start_prices_df['openPrice']-1
    #计算IC
    ic = factor_Series.corr(revenue_Series,method='spearman')
    return ic
```

```python
'''
历史因子IC统计
'''
ic_df = pd.DataFrame()
for trade_date in trade_dates[:-1]:
    #获取因子参考日
    reference_date = reference_dates[trade_dates.index(trade_date)]
    #期初股票池（选股时需按技术指标过滤）
    start_stocks = universe.preview(reference_date, skip_halted=True)
    start_stocks = filter_stocks(reference_date,start_stocks)
    #期末股票池（剔除由于停牌不能卖出的股票）
    next_trade_date = trade_dates[trade_dates.index(trade_date)+1]
    end_stocks = universe.preview(next_trade_date, skip_halted=True)
    #股票池取交集（计算IC需保证前后股票池一致）
    stocks = [stock for stock in start_stocks if stock in end_stocks]
    for factor in factors:
        factor_Series = DataAPI.MktStockFactorsOneDayGet(secID=stocks,tradeDate=reference_date, \
                                                         field=['secID',factor]).set_index('secID')[factor]
        ic_df.loc[trade_date,factor] = calculate_ic(trade_date,next_trade_date,factor_Series)
```

长期看，“股权回报率”与“净利润增长率”IC均应大于0，因为这两项指标越高，说明公司业绩越好，也应有更高的涨幅，但由于选股与数据发布有时间差，导致这些因素可能已被市场反应甚至过度反应，从而股价在选股后出现回落，导致IC在某些月份表现出负值。虽然我们在选股时已经利用RSI剔除了部分超涨股票，在计算权重时，为保证权重IC为正，设定IC均值下限为0.001。

```python
'''
基于滚动3个月IC加权构造复合因子Composite
'''
#确定各期因子权重
for trade_date in trade_dates[3:-1]:
    #计算均值的起始时间为3个月前
    start_date =  trade_dates[trade_dates.index(trade_date)-3]
    end_date = trade_dates[trade_dates.index(trade_date)-1]
    #计算滚动3个月IC均值
    meanIC_roe = ic_df.loc[start_date:end_date,'ROE'].mean()
    meanIC_npg = ic_df.loc[start_date:end_date,'NetProfitGrowRate'].mean()
    #记录各期因子权重（非正值置为0.001）
    m_roe = max(meanIC_roe,0.001)
    m_npg = max(meanIC_npg,0.001)
    ic_df.loc[trade_date,'w_roe'] = m_roe/(m_roe+m_npg)
    ic_df.loc[trade_date,'w_npg'] = m_npg/(m_roe+m_npg)
ic_df.dropna(axis=0,inplace=True)
```

在调仓期初，按照复合因子值从高到低排序，由前20名股票构造顶层组合，后20股票构造底层组合，如果因子有效，预期结果应是：顶层组合有超额正收益，底层组合有超额负收益，且长期来看，顶层组合、沪深300和底层组合的净值表现依次减弱。

```python
#构造数据框用于记录组合收益
revenue_df = pd.DataFrame()
for trade_date in trade_dates[3:-1]:
    #获取调仓日对应因子参考日
    reference_date = reference_dates[trade_dates.index(trade_date)]
    #获取下一调仓日
    next_trade_date = trade_dates[trade_dates.index(trade_date)+1]
    #获取参考日股票池
    stocks = universe.preview(reference_date, skip_halted=True)
    stocks = filter_stocks(reference_date,stocks)
    #获取当期因子值
    factors_df = DataAPI.MktStockFactorsOneDayGet(secID=stocks,tradeDate=reference_date, \
                                                  field= ['secID']+factors).set_index('secID')
    #因子数据标准化
    for factor in factors:
        factors_df[factor] = pd.Series(standardize(factors_df[factor].to_dict()))
    #计算股票池的当期复合因子
    factors_df['Composite'] = ic_df.loc[trade_date,'w_roe']*factors_df['ROE']+ \
                              ic_df.loc[trade_date,'w_npg']*factors_df['NetProfitGrowRate']
    #计算复合因子的IC
    ic_df.loc[trade_date,'Composite'] = calculate_ic(trade_date,next_trade_date,factors_df['Composite'])
    #根据复合因子排名构建当期股票多空组合
    top_list = factors_df.sort(columns=['Composite']).index[-n:].tolist()
    bottom_list = factors_df.sort(columns=['Composite']).index[:n].tolist()
    #剔除调仓日停牌的股票后重新构建当期股票多空组合
    start_pool = universe.preview(trade_date, skip_halted=True)
    end_pool = universe.preview(next_trade_date, skip_halted=True)
    top_list = list(set(top_list)&set(start_pool)&set(end_pool))
    bottom_list = list(set(bottom_list)&set(start_pool)&set(end_pool))
    #计算沪深300的调仓期收益
    hs300_df = DataAPI.MktIdxdGet(ticker='000300',beginDate=trade_date,endDate=next_trade_date,field=['openIndex'])
    revenue_df.loc[trade_date,'hs300'] =  hs300_df['openIndex'].iloc[-1]/hs300_df['openIndex'].iloc[0]-1
    #计算多头组合的调仓期收益
    start_df = DataAPI.MktEqudAdjGet(secID=top_list,tradeDate=trade_date,field=['secID','openPrice'])
    end_df = DataAPI.MktEqudAdjGet(secID=top_list,tradeDate=next_trade_date,field=['secID','openPrice'])
    revenue_df.loc[trade_date,'top'] = (end_df['openPrice']/start_df['openPrice']-1).mean()
    #计算空头组合的调仓期收益
    start_df = DataAPI.MktEqudAdjGet(secID=bottom_list,tradeDate=trade_date,field=['secID','openPrice'])
    end_df = DataAPI.MktEqudAdjGet(secID=bottom_list,tradeDate=next_trade_date,field=['secID','openPrice'])
    revenue_df.loc[trade_date,'bottom'] = (end_df['openPrice']/start_df['openPrice']-1).mean()
```

```python
#计算组合单位净值
cost = 0.0015
netvalues_df = pd.DataFrame()
netvalues_df['hs300'] = (1+revenue_df['hs300'].shift(1).fillna(0)).cumprod()
netvalues_df['top'] = (1+revenue_df['top'].shift(1).fillna(0)-cost).cumprod()
netvalues_df['bottom'] = (1+revenue_df['bottom'].shift(1).fillna(0)-cost).cumprod()
```

```python
#统计组合表现
performance_df = pd.DataFrame()
for portfolio in netvalues_df.columns:
    #计算年化平均收益
    performance_df.loc[portfolio,'Annual'] = 12*revenue_df[portfolio].mean()
    #计算年化波动率
    performance_df.loc[portfolio,'StdError'] = np.sqrt(12)*revenue_df[portfolio].std()
    #计算超额收益
    excess_return = revenue_df[portfolio]-revenue_df['hs300']
    #计算年化Alpha
    performance_df.loc[portfolio,'Alpha'] = 12*excess_return.mean()
    #计算组合Beta
    performance_df.loc[portfolio,'Beta'] = revenue_df[portfolio].cov(revenue_df['hs300'])/revenue_df['hs300'].var()
    #计算最大回撤
    performance_df.loc[portfolio,'Maxdown']= max(1-netvalues_df[portfolio]/np.maximum.accumulate(netvalues_df[portfolio]))
    #计算年化跟踪误差
    performance_df.loc[portfolio,'TrackError'] = np.sqrt(12)*excess_return.std()    
#计算夏普比率
performance_df['Sharp'] = (performance_df['Annual']-0.03)/performance_df['StdError']
#计算信息比率
performance_df['IR'] = performance_df['Alpha']/(performance_df['TrackError']+0.000001)
performance_df.applymap(lambda x:round(x,3))
```

组合业绩显示：顶层组合在各项指标均给出了最佳表现，而底层组合与基准的年化收益相等，但其它指标落后基准。

```python
'''
绘制组合超额收益图
'''
top_ex = revenue_df['top']-revenue_df['hs300']
bottom_ex = revenue_df['bottom']-revenue_df['hs300']
x = np.arange(len(revenue_df))
width = 0.4
x_labels = [date[0:7] for date in revenue_df.index]
fig,axes = plt.subplots(1,1,figsize=(16,4))
axes.set_title(u'组合月度超额收益率',fontproperties=font,fontsize=18)
axes.set_xlim([-1,len(revenue_df)])
axes.set_xticks(x)
axes.set_xticklabels(x_labels,rotation=45,fontsize=12)
axes.set_ylim([-0.1,0.12])
axes.set_ylabel(u'收\n益\n率',fontproperties=font,fontsize=14,rotation=0,labelpad=12)
axes.bar(x-width,top_ex,width,color='red',label=u'顶层组合超额收益')
axes.bar(x,bottom_ex,width,color='green',label=u'底层组合超额收益')
axes.legend(prop=font,fontsize=16)
plt.show()
```

月度超额收益显示：大部分时间内，顶层组合超额收益为正，底层组合超额收益为负，符合预期。从超额收益率这一角度，复合因子选股有效。

从复合因子权重构成看，大部分时间内，股权回报率对复合因子起更重要的作用。

因子月度IC显示：大部分时间内，股权回报率IC>复合因子IC>净利润增长率IC，从这一角度看似乎按照股权回报率因子选股更好。但股权回报率因子也有表现极端差的时候，比如2019年2月，2019年11月和2020年6月都出现了极端负值，此时复合因子就起到了中和作用，所以复合因子的表现更稳定。

```python
'''
因子IC表现统计
'''
icpfm_df = pd.DataFrame()
for factor in ['ROE','NetProfitGrowRate','Composite']:
    icpfm_df.loc[factor,'Mean'] = ic_df[factor].mean()
    icpfm_df.loc[factor,'Std'] = ic_df[factor].std()
icpfm_df['ICIR'] = icpfm_df['Mean']/icpfm_df['Std']
icpfm_df.applymap(lambda x:round(x,3))
```

从净值曲线看，顶层组合>沪深300>底层组合，符合预期，从这一角度说明了选股策略有效。其中顶层组合是我们实际投资的组合，相对收益上，取得了明显优于基准的回报；绝对收益上，用时19个月实现了净值翻倍。
