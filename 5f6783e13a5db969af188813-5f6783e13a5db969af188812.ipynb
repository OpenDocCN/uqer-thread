```python
# 行业分类标准 

# DataAPI.IndustryGet(industryVersion=u"",industryVersionCD=u"010303",industryLevel=u"1",isNew=u"1",prntIndustryID=u"",industryName="",industryID=u"",field=u"",pandas="1")

"""
行业分类标准
[通联数据] - DataAPI.IndustryGet
DataAPI.IndustryGet(industryVersion=u"SW",industryVersionCD=u"",industryLevel=u"1",isNew=u"1",prntIndustryID=u"",field=u"",pandas="1")
名称	              类型	 描述
industryVersion 	str	    行业分类标准。SW- 申万行业分类,industryVersion、industryVersionCD至少选择一个
industryVersionCD	list	行业分类标准数字编码。010301-证监会行业V2012、010303-申万行业、010308-中证行业,可以是列表,industryVersion、industryVersionCD至少选择一个
                            （！补充内容：010306-恒生行业  010314-中证行业分类（2016版）  010317-中信行业分类  # 010319-战略性新兴产业(2018) ， 没有和 同花顺上(66个) 相同的分类；）
industryLevel	    int	    输入数字如1,2,3,4，可指定查询的第几级行业,可以是列表,可空
isNew	            int	    是否最新，1表示是，0表示否,可空
prntIndustryID	    list	上级行业数字编码，通联赋予，输入可查询该分类下所有子分类情况,可以是列表,可空
field	            list	所需字段,可以是列表,可空
pandas	            str	    1表示返回 pandas data frame，0表示返回csv,可空
"""

def hangyefenlei(pd_xyfl):
    pd_xyfl.rename(columns={'industryVersionCD':'1行业分类标准数字编码',
                             'industryVersion':'2行业分类标准(字母缩写)',
                             'industry':'3行业分类标准名称',
                             'industryID':'4行业划分数字编码，通联赋予',
                             'industrySymbol':'5行业划分(编制机构)',
                             'industryName':'6行业名称',
                             'industryLevel':'7行业划分所属级别',
                             'isNew':'8是否最新',
                             'indexSymbol':'9对应行业指数代码(编制机构)',
                             'prntIndustryID':'10上级行业数字编码，通联赋予，与industryID对应',
                             'beginDate':'11分类生效日期',
                             'endDate':'12分类截止日期',
                             'updateTime':'13最近的一次更新时间'},inplace=True)
    return pd_xyfl

def xyfl(*args):
    if len(args)==1:
        _prntIndustryID=args[0]
    else:
        return
    
    if _prntIndustryID=='1':
        pd_return=pd_xyfl[pd_xyfl['7行业划分所属级别']==1]
        pd_return['ID']=range(1,pd_return.shape[0]+1)
        pd_return.set_index('ID',inplace=True)
        return pd_return
    elif _prntIndustryID=='all':
        return pd_xyfl
    else: # 4行业划分数字编码，通联赋予
        pd_return=pd_xyfl[(pd_xyfl['7行业划分所属级别']==1) & (pd_xyfl['4行业划分数字编码，通联赋予']==_prntIndustryID)]
        return xyfl_sub(pd_return,_prntIndustryID)

def xyfl_sub(pd_return,_prntIndustryID):
    _pd=pd_xyfl[(pd_xyfl['7行业划分所属级别']!=1) & (pd_xyfl['10上级行业数字编码，通联赋予，与industryID对应']==_prntIndustryID)] 
    if _pd.shape[0]>0:
        for item_pd in _pd['4行业划分数字编码，通联赋予'].tolist():  
            pd_return=pd_return.append(pd_xyfl[pd_xyfl['4行业划分数字编码，通联赋予']==str(item_pd)], ignore_index=True)
            pd_return=xyfl_sub(pd_return,str(item_pd))
    return pd_return

pd_xyfl=hangyefenlei(DataAPI.IndustryGet(industryVersionCD=u"010303",isNew=u"1",pandas="1")) # 010303-申万行业
```
