```python
def show_pan(account,ezq):
    #这个函数用来返回一些大盘信息，根据是否ezq来判断加工不加工返回的信息
    global pan_zhi_today,pan_zhi_yesterday,re
    pan_name = '000300'#沪深300指数

    today = account.current_date
    yesterday = account.previous_date
    
    pan_field = ['closeIndex', 'highestIndex','lowestIndex', 'turnoverVol','CHG']
    
    pan_zhi_today = DataAPI.MktIdxdGet(ticker= pan_name ,tradeDate = today , field=pan_field)
    pan_zhi_yesterday = DataAPI.MktIdxdGet(ticker= pan_name ,tradeDate = yesterday , field=pan_field)
    
    today_index = pan_zhi_today['closeIndex'][0]
    today_difference = (pan_zhi_today['highestIndex']-pan_zhi_today['lowestIndex'])[0]
    today_turnoverVol =  pan_zhi_today['turnoverVol'][0]
    today_CHG = pan_zhi_today['CHG'][0]
    
    yesterdat_index = pan_zhi_yesterday['closeIndex'][0]
    yesterdaty_difference = (pan_zhi_yesterday['highestIndex']-pan_zhi_yesterday['lowestIndex'])[0]
    yesterday_turnoverVol =  pan_zhi_yesterday['turnoverVol'][0]
    yesterday_CHG = pan_zhi_yesterday['CHG'][0]

    index_change = today_index - yesterdat_index
    
    difference_change = today_difference/yesterdaty_difference
    CHG_change = today_CHG - yesterday_CHG
    turnoverVol_change = today_turnoverVol - yesterday_turnoverVol
    
    if ezq:
        is_index_up = five_return(index_change)
        is_turnoverVol_change = five_return(turnoverVol_change)
        is_CHG_change = five_return(CHG_change)
        return [is_index_up,is_turnoverVol_change,is_CHG_change]
    re =  [today_index,index_change,difference_change,turnoverVol_change,today_CHG,CHG_change]
    
    return re
    #返回今天的指数，指数变化，最大差值变化，交易量变化，今日涨幅，涨幅变化
def five_return(w):
    #这是个辅助函数，方便我偷懒用。
    if w>.05:
        re = 2
    elif w>.02:
        re = 1
    elif w>-.02:
        re = 0
    elif w>-.05:
        re = -1
    else:
        re = -2
    return re
```

```python
import random
class Learner:
    def __init__(self , stock , learning = False , alpha = .5 , epsilon =1.0 ):
        
        self.ezq = True #第一个Qlearn就默认为True吧。其实可以用线性回归的方法，也可以做增强学习。线性回归可以的话，神经网络自然也可以搭起来（以后再说吧……写着费力。）
        
        self.stock = stock #这次增强学习的是那只股票呢？
        self.learning = learning #是否需要学习，设定为False的时候是debug用的。也就是随机情况。
        self.Q = dict()
        self.epsilon = epsilon
        self.alpha = alpha
        
        self.base_reward = 10 #基准奖励判定设定为，用这个值乘以涨跌来判断奖励。这个值越大学习机越敏感，越小越迟钝。
        
        self.state = None #初始状态设定为None，我们把状态的更新放在后面。
        self.count_reward = 0
        self.turn = 0 #the turn of this Learning now
        self.num_of_learn = 0 #the number of times of learn now
        self.learns = {} #get state of every turns
        
        self.last_state = None
        self.last_action = None
        
        self.all_reward_now = 0 #计算当前的总奖励，用于分析不同股票优劣
        self.all_reward_last = 0
        
        self.action = ['buy','sale',None] #我们选择的动作，这里我们对股票进行买，卖和啥事也不干3个动作。
        
    def reset(self,testing = False):
        
        # If 'testing' is True, set epsilon and alpha to 0
        if testing:
            self.epsilon = 0
            self.alpha = 0
        
        self.all_reward_now = self.count_reward
        
        #let the epsilon change with time go,也可以成为自大控制……epsilion越大越不自信。
        if self.all_reward_now <= self.all_reward_last :
            self.epsilon += .01
        elif self.all_reward_now > self.all_reward_last :
            self.epsilon *= .90
        
    def build_state(self,account_information,stock):
        
        pan_zhi = account_information
        # pan_zhi = show_pan(account,self.ezq)
        gu_zhi = self.make_stock_information(stock)
        #stock = ?
        raw_stock_state = pan_zhi + gu_zhi
        raw_state = tuple(raw_stock_state)
        
        self.createQ(raw_state)
        return raw_state
    
    def get_maxQ(self, state):
        #找到q最大的一个动作
        max_action = None
        max_action_value = .0
        
        for action in self.Q[state].keys():
            if self.Q[state][action] > max_action_value:
                max_action = action
                max_action_value = self.Q[state][action]

        maxQ = (max_action , max_action_value)

        return maxQ
    def createQ(self, state):
        #创造Q的状态下的动作，这里分为3个，买，卖和不动。
        if not state in self.Q.keys():
            # print '*********',type(state)
            self.Q[state] = dict()
            
            for action in self.action:
                self.Q[state][action] =.0
            # self.Q[state][None] =.0
            # self.Q[state]['buy'] =.0
            # self.Q[state]['sale'] = .0
            # self.Q[state]['do_nothing'] = .0
        return
        
    def choose_action(self, state):
        #随机数大于探索率，则选择Qtable中最优的，否则随机选。
        next_random_action = random.choice( self.action )
        
        if self.learning and random.random() > self.epsilon:
                action = self.get_maxQ(state)[0]
        else:
            action = random.choice( self.action )
            
        return action

    def learn(self, state, action, reward):
        #从状态和动作中学习学习，学习效率为alpha，即alpha的比例学习，1-alpha的比例记忆以前的经验。
        q = self.Q[state][action]
        q = (1-self.alpha)*q + self.alpha * reward
        self.Q[state][action] = q
        return
        
    def update(self,account_information,stock):
        #每次要更新一下几个数，不过由于股票涨跌只有第二天能看到，所以这个更新的reward是更新的上一天的才能看到……
        state = self.build_state(account_information,stock)          # Get current state
        self.createQ(state)                 # Create 'state' in Q-table
        action = self.choose_action(state)  # Choose an action
        
        reward = self.make_stock_reward(self.stock) # Receive a reward
     
        try :
            self.learn(self.last_state, self.last_action, reward)   # 用昨天的状态，行为和今天的奖励来学习一个。
        except:
            pass
            # print "can't learn from last state or action"
            # print 'last state:',state
            # print 'last_action:',action

        self.last_state = state
        self.last_action = action
        
        self.turn += 1
        self.count_reward += reward
        self.learns[self.num_of_learn] =( self.turn,self.count_reward )
        return
    
    def make_all_stock_information(self,all_stock_information):
        #这个函数是用来处理个股信的准备工作。
        [price_change,turnoverVol_change,Difference_today,Difference_yesterday] = all_stock_information
        
        self.all_price_change = price_change
        self.all_turnoverVol_change = turnoverVol_change
        self.all_Difference_today = Difference_today
        self.all_Difference_yesterday = Difference_yesterday
        
    def make_stock_information(self,stock):
        #这个函数是用来处理个股信息的，把个股信息从预处理信息中提出来。
        price_change = self.all_price_change[stock]
        turnoverVol_change = self.all_turnoverVol_change[stock]
        Difference_today = self.all_Difference_today[stock]
        Difference_yesterday = self.all_Difference_yesterday[stock]
        
        Difference_change = (Difference_today-Difference_yesterday)/Difference_yesterday
        
        if self.ezq:#如果是简单版的Qlearn,不使用线性回归，直接定义几个状态。
            # 价格涨了吗？状态分别为大涨，涨，小涨，不需要管和跌。
            if price_change>.2:
                is_price_up = 3
            elif price_change > .1:
                is_price_up = 2
            elif price_change > .02:
                is_price_up = 1
            elif price_change > -.02:
                is_price_up = 0
            else:
                is_price_up = -1
                
            #交易量涨了吗？
            if turnoverVol_change > .05:
                is_turnoverVol_up = 1
            elif turnoverVol_change > -.05:
                is_turnoverVol_up = 0
            else:
                is_turnoverVol_up = -1
            
            #日最大差涨了吗？
            if Difference_change > .05:
                is_difference_change = 1
            elif Difference_change > -.05:
                is_difference_change = 0
            else:
                is_difference_change = -1
                
            return [is_price_up,is_turnoverVol_up,is_difference_change]
            
        price_up = price_change>0
        price_up_alot = price_change>.1
        turnoverVol_change = turnoverVol_change

        return [price_change,turnoverVol_change,Difference_today,Difference_yesterday]
    
    def make_stock_reward(self,stock):
        #把股票(今天减昨天/昨天)和昨天做出的买卖行为比较，来判断reward
        
        yestertoday_action_reward = self.all_price_change[stock] * self.base_reward
        
        return yestertoday_action_reward
    
    def get_w(self):
        #这个函数是未来用于线性回归而非简单的判断的时候，用来返回参数以debug用的。
        pass
```
