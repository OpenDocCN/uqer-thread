# 涨跌停股票统计

```python
import pandas as pd

today = Date.todaysDate()
cal = Calendar('China.SSE')
yesterday = cal.advanceDate(today, '-1B',  BizDayConvention.Following).strftime('%Y%m%d')

allSecList = DataAPI.EquGet(equTypeCD = 'A', field = ['secID'])['secID'].tolist()
precls = {}
uplimit, dnlimit = [], []
for i in range(0, len(allSecList), 200):
    sub = allSecList[i:min(len(allSecList), i+200)]
    df_precls = DataAPI.MktEqudGet(secID = sub, beginDate = yesterday, endDate = yesterday, field = 'secID,closePrice')
    df_lasprx = DataAPI.MktTickRTSnapshotGet(securityID = sub, field = 'shortNM,lastPrice,bidBook,askBook')
    for j in range(len(df_precls)):
        precls[df_precls.at[j,'secID']] = df_precls.at[j,'closePrice']
    for j in range(len(df_lasprx)):
        if df_lasprx.at[j,'lastPrice'] > 0:
            sec = df_lasprx.at[j,'ticker']+'.'+df_lasprx.at[j,'exchangeCD']
            ret = df_lasprx.at[j,'lastPrice'] / precls[sec] - 1
            if df_lasprx.at[j,'bidBook_volume1'] == 0 and df_lasprx.at[j,'askBook_volume1'] > 0:
                dnlimit.append([df_lasprx.at[j,'shortNM'], sec, precls[sec], df_lasprx.at[j,'lastPrice'], ret])
            if df_lasprx.at[j,'askBook_volume1'] == 0 and df_lasprx.at[j,'bidBook_volume1'] > 0:
                uplimit.append([df_lasprx.at[j,'shortNM'], sec, precls[sec], df_lasprx.at[j,'lastPrice'], ret])
```

```python
name, sec, precls, latprx, ret = zip(*uplimit)
df_uplimit = pd.DataFrame({'name': name, 'sec': sec, 'preClosePrice': precls, 'latestPrice': latprx, 'return': ret}).sort(columns='sec').reset_index()
df_uplimit = df_uplimit.loc[:, ['name', 'sec','preClosePrice','latestPrice','return']]
print '涨停股票数量：', len(uplimit)
df_uplimit
```

```python
name, sec, precls, latprx, ret = zip(*dnlimit)
df_dnlimit = pd.DataFrame({'name': name, 'sec': sec, 'preClosePrice': precls, 'latestPrice': latprx, 'return': ret}).sort(columns='sec').reset_index()
df_dnlimit = df_dnlimit.loc[:, ['name', 'sec','preClosePrice','latestPrice','return']]
print '跌停股票数量：', len(dnlimit)
df_dnlimit
```
