# 基于奇异谱分析的均线择时研究

```python
from __future__ import division
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model

#嵌入
def getWindowMatrix(inputArray, t, m):
    temp = []
    n = t-m+1
    #t-m+1为矩阵的列的大小
    for i in range(n):
        temp.append(inputArray[i:i+m])
        #0到m-1为第一行，即m个数字为一行，最后一行首个数字为t-m,最后一个为t-1,对应时间序列里面的x(t),原因是range(n)从0开始，故以上均减一处理
    WindowMatrix = np.array(temp)
    return WindowMatrix

#svd分解并重构矩阵
def SVDreduce(WindowMatrix):
    u,s,v = np.linalg.svd(WindowMatrix)
    m1, n1 = u.shape
    m2, n2 = v.shape
    index = s.argmax()
    u1 = u[:, index]
    v1 = v[index, :]
    u1 = u1.reshape((m1, 1))
    v1 = v1.reshape((1, n2))
    value = s.max()
    newMatrix = value * (np.dot(u1, v1)) #重构矩阵
    return newMatrix

#对角线平均法重构序列
def recreateArray(newMatrix, t, m):
    ret = []
    n = t-m+1
    for p in range(1, t+1):
        if p<m:
            alpha = p
        elif p > t-m+1:
            alpha = t-p+1
        else:
            alpha = m
        sigma = 0
        for j in range(1, m+1):
            i = p-j+1
            if i>0 and i<n+1:
                sigma = sigma + newMatrix[i-1][j-1]
        ret.append(sigma/alpha)
    return ret

#按不同序列，不同窗口计算大小计算SSA
#得到新的重构之后的时间序列
def SSA(inputArray,t,m):
    WindowMatrix = getWindowMatrix(inputArray,t,m)
    newMatrix = SVDreduce(WindowMatrix)
    newArray = recreateArray(newMatrix, t, m)
    return newArray

#MA均线计算
def MA(inputArray, m):
    output = []
    length = len(inputArray)
    for i in range(m):
        output.append(0)
    for i in range(m, length):
        temp = inputArray[i-m:i]
        output.append(temp.sum()/m)
    return output
```

```python
m = 5
t = 2*m
netValue_ssa = 1
netValue_ma = 1
netValue_hold = 1
closeIndex = np.array(data['closeIndex'])
openIndex = np.array(data['openIndex'])
datelist = np.array(data['tradeDate'])

flag_ssa = False
flag_ma = False

strategy_ssa = {}
strategy_ma = {}
strategy_hold = {}
length = len(closeIndex)
startHold_ssa = [] #记录多头当期净值以及成本价
startHold_ma = [] #hold表示简单的买入并持有之

cost_hold = openIndex[t]
for i in range(t, length):
    yesterday = datelist[i-1]
    today = datelist[i]
    inputArray = closeIndex[:i]
    inputArray = inputArray[-t:]
    SSA5 = SSA(inputArray,t,m)
    MA5 = MA(inputArray,m)
    valueYesterday = closeIndex[i-1]
    valueToday = openIndex[i]
    ########################################################## ssa
    if len(startHold_ssa) != 0: #持有头寸
        netValue_ssa = (startHold_ssa[0]) * (valueYesterday/startHold_ssa[1])
    else: #空仓时
        netValue_ssa = netValue_ssa
    strategy_ssa.update({yesterday:netValue_ssa})
    
    #strategy_ssa判断，价格线刚刚穿破ssa5时买入
    if flag_ssa == False and valueYesterday>SSA5[-1] and closeIndex[i-2]<SSA5[-2]:
        flag_ssa = True
        startHold_ssa = [netValue_ssa, valueToday]
        #只有买入卖出时startHold才会变,即此处的净值与成本价都是指刚买入时的值
    
    #strategy_ssa判断卖出
    if flag_ssa == True and valueYesterday<SSA5[-1] and closeIndex[i-2]>SSA5[-2]:
        netValue_ssa = (startHold_ssa[0])*(valueToday/startHold_ssa[1])
        #因为卖出后再以新的价格买入时其初始净值会改变
        flag_ssa = False
        startHold_ssa = []
    ######################################################### ma
    if len(startHold_ma) != 0:
        netValue_ma = (startHold_ma[0]) * (valueYesterday/startHold_ma[1])
    else:
        netValue_ma = netValue_ma
    strategy_ma.update({yesterday:netValue_ma})
    
    #MA 判断买入
    if flag_ma == False and valueYesterday>MA5[-1] and closeIndex[i-2]<MA5[-2]:
        flag_ma = True
        startHold_ma = [netValue_ma, valueToday]
    # MA判断卖出
    if flag_ma == True and valueYesterday<MA5[-1] and closeIndex[i-2]>MA5[-2]:
        netValue_ma = (startHold_ma[0])*(valueToday/startHold_ma[1]) 
        flag_ma = False
        startHold_ma = []  
    ####################################################### hold 买入并持有之
    netValue_hold = valueYesterday/cost_hold
    strategy_hold.update({yesterday:netValue_hold})
    
```

```python
from sklearn import linear_model

def LoopbackTest(data, combination):
    netValue = 1
    numOfTrading = 0
    if 'closeIndex' in data.columns:
        closeIndex = np.array(data['closeIndex'])
        openIndex = np.array(data['openIndex'])
    elif 'closePrice' in data.columns:
        closeIndex = mp.array(data['closePrice'])
        openIndex = np.array(data['openPrice'])
    dateList = np.array(data['tradeDate'])
    flag = False
    strategy_name = 'strategy'+'('
    for i in range(len(combination)):
        strategy_name = strategy_name+str(combination[i])+','
    strategy_name = strategy_name[:-1]+')'
    strategy = {} #记录策略每天净值
    length = len(closeIndex)
    startHold = [] #记录多头当期成本价及当期净值
    endHold = []
    t_max = 2*max(combination) #最大窗口ssa线需要的最小天数
    ###########################################################
    ############################################################
    
    #############################################################
    ############################################################# 每天的交易
    #每天开盘后以昨天的收盘价判断，但以今天的开盘价成交
    for i in range(t_max, length):
        yesterday = dateList[i-1]
        today = dateList[i]
        valueYesterday = closeIndex[i-1]
        valueToday = openIndex[i]
        
        #计算当前净值（截止昨天收盘为止，即按昨天收盘价计算的净值，每天更新一次）
        if len(startHold) != 0:
            if valueToday == 0:
                netValue = netValue
            else:
                netValue = startHold[0]*(valueYesterday/startHold[1])
        else:
            netValue = netValue
        strategy.update({yesterday:netValue})
        
        SSA_score = []
        for m in combination:
            t = 2*m
            inputArray = closeIndex[:i]
            inputArray = inputArray[-t:]
            SSAx = SSA(inputArray, t, m)
            if SSAx[-1] >= SSAx[-2]:
                SSA_score.append(1)
            else:
                SSA_score.append(-1)
        score = sum(SSA_score)
        
        #判断买入
        if score>0 and flag==False and valueToday!=0:
            flag = True
            netValue = netValue * 0.998
            numOfTrading += 1
            startHold = [netValue, valueToday]
        
        #判断卖出
        if flag==True and score<0:
            flag = False
            netValue = (startHold[0]) * (valueToday/startHold[1])
            netValue = netValue * 0.998
            numOfTrading += 1
            startHold = [] #清仓
    return {strategy_name:(strategy, numOfTrading)}

#年化利率
def AnnualizedReturns(pStart, pEnd, n):
    return (pEnd/pStart)**(250/n) - 1

#最大回撤（输入为list或则array）
def MaxDrawdown(strategy):
    length = len(strategy)
    drawdown = []
    py = strategy[0]
    for i in range(1, length):
        px = strategy[i]
        py = max(strategy[:i])
        drawdown.append(1 - px/py)
    return max(drawdown)

#斜率（输入为list或array）
def Slope(strategy):
    num = len(strategy)
    x = np.mat(np.arange(num))
    x = x.T
    y = strategy - 1
    clf = linear_model.LinearRegression()
    clf.fit(x, y)
    slope = clf.coef_[0]
    return slope

#策略表现,对应上述LoopbackTest生成得字典
def Performance(strategy):
    strategyName = strategy.keys()[0]
    netValueDict = strategy[strategyName][0] #净值及对应日期
    netValueArray = np.array(pd.Series(strategy[strategyName][0]))
    numOfTrading = strategy[strategyName][1]
    pStart = 1
    pEnd = netValueArray[-1]
    n = len(netValueArray)
    slope = Slope(netValueArray)
    maxdrawdown = MaxDrawdown(netValueArray)
    annReturns = AnnualizedReturns(pStart, pEnd, n)
    performance = {}
    performance.update({u'拟合斜率':slope})
    performance.update({u'最终净值':pEnd})
    performance.update({u'年化收益率':annReturns})
    performance.update({u'最大回撤':maxdrawdown})
    performance.update({u'交易次数':numOfTrading})
    return {strategyName:performance}

#自动测试并生成表格
def AutoTest(data,parameter_dict):
    ret = {}
    for num_of_line in parameter_dict.keys():
        combinations = parameter_dict[num_of_line]
        performanceDF = pd.DataFrame()
        for combination in combinations:
            tempStrategy = LoopbackTest(data, combination)
            performance = Performance(tempStrategy)
            key = performance.keys()[0]
            performanceDF[key] = pd.Series(performance[key])
        ret.update({num_of_line:performanceDF.transpose()})
        print num_of_line,'条均线组合测试完成'
    return ret
            
```
