```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import datetime as dt
import CAL.PyCAL as C_cal # CAL provides classes and functions dealing with date, time, and calendars. 

sns.set_style('white')
pd.set_option('display.width', 200)
cal = C_cal.Calendar('China.SSE')
```

```python
start_date = '20180801' # 开始日期
end_date = (dt.datetime.today() - dt.timedelta(days=1)).strftime('%Y%m%d') # 截止日期

ticker,ticker1='600674', '600886'
# 获取股票的基本信息，包含股票交易代码及其简称、股票类型、上市状态、上市板块、上市日期等；上市状态为最新数据，不显示历史变动信息。
full_name=DataAPI.EquGet(ticker=ticker,field='secFullName',pandas="1").iloc[0,0]
Short_Name=DataAPI.EquGet(ticker=ticker,field='secShortName',pandas="1").iloc[0,0]
full_name1=DataAPI.EquGet(ticker=ticker1,field='secFullName',pandas="1").iloc[0,0]
Short_Name1=DataAPI.EquGet(ticker=ticker1,field='secShortName',pandas="1").iloc[0,0]

print(full_name+'|'+Short_Name+'|'+ticker+'\n'+
      full_name1+'|'+Short_Name1+'|'+ticker1+'\n'+
      '开始日期:'+(start_date[0:4]+'年'+start_date[4:6]+'月'+start_date[6:8]+'日')+'    截止日期:'+(end_date[0:4]+'年'+end_date[4:6]+'月'+end_date[6:8]+'日')+'\n')

"""
获取沪深A股和B股前复权日行情信息，包含前复权昨收价、前复权开盘价、前复权最高价、前复权最低价、前复权收盘价。
DataAPI.MktEqudAdjGet(secID=u"",ticker=u"",tradeDate=u"20150513",beginDate=u"",endDate=u"",isOpen="",field=u"",pandas="1")

前复权就是以目前股价为基准，保持现有价位不变，缩减以前价格，把除权前的K线向下平移，使图形吻合，保持股价走势的连续性。简单说就是把除权前的价格按现在的价格换算过来，复权后现在价格不变，以前的价格减少。

前复权公式：复权后价格=(复权前价格-现金红利)/(1+流通股份变动比例)
"""
data = DataAPI.MktEqudAdjGet(ticker=ticker, beginDate=start_date, endDate=end_date,pandas="1")
data1 = DataAPI.MktEqudAdjGet(ticker=ticker1, beginDate=start_date, endDate=end_date,pandas="1")

liangtou =pd.DataFrame()
liangtou['川投股价'] = data.closePrice
liangtou['国投股价'] = data1.closePrice
liangtou['川投/国投股价比例'] =  data.closePrice / data1.closePrice

liangtou[['川投股价','国投股价']].plot(figsize=(16,5),grid = True)
plt.legend([u'川投股价',u'国投股价'], prop=C_cal.font)

liangtou[['川投/国投股价比例']].plot(figsize=(16,5),grid = True)
plt.legend([u'川投/国投股价比例'], prop=C_cal.font)
```
