## 1. 徐翔概念股
-----

十一月初，总舵主徐翔被警方带走，曾引发相关股票巨震。

一个月后，我们回顾徐翔概念股在那一段时间走势，具体如何呢？

```python
import pandas as pd

# 泽熙投资重仓股票
zexi = pd.read_csv('zexi.csv',encoding='GB18030')
zexi = zexi[['secID','secShortName']]
zexi
```

按照社区 @陈亚龙 大神的帖子，如上表整理了一个泽熙重仓的股票列表，原帖地址：
https://uqer.datayes.com/community/share/56380a9df9f06c713ddfe922

```python
import seaborn
from CAL.PyCAL import *

zexi_data = pd.DataFrame()

for sec in zexi.secID:
    fields = 'secID,tradeDate,closePrice'
    tmp = DataAPI.MktEqudGet(secID=sec, beginDate='20151030', endDate='20151126', field=fields, pandas='1')
    zexi_data['tradeDate'] = tmp.tradeDate
    zexi_data[sec] = tmp.closePrice/tmp.closePrice[0]-1.0
    
hs300 = DataAPI.MktIdxdGet(ticker='000300', beginDate='20151030', endDate='20151126', field='closeIndex', pandas='1')
zexi_data['hs300'] = hs300.closeIndex/hs300.closeIndex[0]-1

zexi_data = zexi_data.set_index('tradeDate')
zexi_data.plot(figsize=(10,6),style='-o')
```

上图中，我们给出了沪深300指数和徐翔概念股从10月30日（徐翔被带走之前一个交易日）开始的走势图，图中纵轴为个股股价相对于10月30日的涨跌幅

从图中可以看到:
- 11月1日徐翔被带走后的两天，泽熙重仓股经历了杀跌的过程，沪深300指数也相应受挫；
- 然而，两天后，指数企稳，泽熙重仓股开始反弹，且反弹力度不小；
- 因此上，徐翔概念股的杀跌给了大家一个观望抄底的良机

## 2. 复星概念股
----
江湖传言，郭广昌昨日同样被警察带走，郭身后是庞大的复星帝国，此次市场会有什么样的反应呢？

```python
# 复星概念股
fosun = pd.read_csv('fosun.csv',encoding='GB18030')
fosun = fosun[['secID','secShortName','holdRank']]
fosun
```

按照泽熙概念股的整理方法，上表列出了复星相关的股票；通过复星高科技、复星医药、复星创投、兴业投资、广信科技、南京钢联等，复星在A股市场控股或大量持股超过20家公司:
- 上表中，第三列 holdRank 表示复星在相应上市公司十大股东中的座次；
- 可以发现，复星作为第一大股东直接控制五家公司，作为第二大股东的有七家；
- 无责任猜测，接下来的几天，复星控股的五家公司可能会碰到黑天鹅，需要谨慎对待；
- 其它复星不是第一大股东的公司，影响应该有限，但也应谨慎

```python
fosun_data = pd.DataFrame()

for sec in fosun.secID[0:5]:
    fields = 'secID,tradeDate,closePrice'
    tmp = DataAPI.MktEqudGet(secID=sec, beginDate='20151111', endDate='20151210', field=fields, pandas='1')
    fosun_data['tradeDate'] = tmp.tradeDate
    fosun_data[sec] = tmp.closePrice/tmp.closePrice[0]-1.0
    
hs300 = DataAPI.MktIdxdGet(ticker='000300', beginDate='20151111', endDate='20151210', field='closeIndex', pandas='1')
fosun_data['hs300'] = hs300.closeIndex/hs300.closeIndex[0]-1

fosun_data = fosun_data.set_index('tradeDate')
fosun_data.plot(figsize=(10,6), style='-o')
```

上图中，我们给出了沪深300指数和复星控股的五家公司从11月11日开始的走势图，图中纵轴为个股股价相对于11月11日的涨跌幅
- 五家公司近期的走势均跟随大盘，且都是Beta放大器；
