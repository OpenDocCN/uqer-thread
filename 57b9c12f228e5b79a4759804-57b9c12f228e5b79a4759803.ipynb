在上一篇[国海证券《因子预测能力初探》uqer实现（一）](https://uqer.datayes.com/community/share/57a3ea90228e5b9b8fa86b0f)中，使用RANK-IC来预测因子是否有稳定的预测能力。但是IC是基于线性相关性的假设，无论是Spearman相关系数还是Pearson相关系数都是基于线性方法求解。当序列存在局部线性关系时，我们可以得到较高的相关系数，但是当我们使用打分法或者回归法取最高或最低值时，很有可能落在非单调区间。所以对于两个相关序列，如果选取相关区间也是一个重要的话题。

## 主要方法
将因子在每一周期（1周）初始将因子值进行排序，并将个股划分为10档，在当期结束后计算该档个股的平均收益（**前瞻收益率**），统计因子每档和平均收益率的关系。简单的说，就是因子的分组收益。

## 好的因子应该有怎样的图形
理论上来说，好的因子在每个分位点的收益应该呈线型的递增或者递减。如果分布比较平均或者杂乱无章，该因子的随机性还是比较大，不是好的因子。

## 与研报结果对比

1. 研报中把2014,2015年作为一组，2016年上半年作为一组。当代码中时间设置和研报中的一致时，得到的图形基本一致。
2. 个人感觉这种方式不是特别妥当，2016年上半年也许根本没有代表性，我们自己可以把参数调整到2016年8月份，验证一下。果然，有些因子就不那么稳定了，如RSI

## 结论
1. 研报中的分组收益没有什么太多有新意的东西。但是分组收益通过图形化展示后，可以更加直观的展示该因子是否稳定。
2. 可以看出有些因子在某些时间段还是比较稳定的，几乎找不到一直稳定的因子。因子在什么条件下会有稳定的IC,可以好好挖掘一下，希望大神们能分享一下相关的资料。

```python
import scipy.stats as st
import pandas as pd

def get_weekend_trading_day_list(begin_day="20140101", end_day="20160801"):
    """获取每周的最后一个交易日"""
    cal_dates = DataAPI.TradeCalGet(exchangeCD=u"XSHG", beginDate=begin_day, endDate=end_day, field="calendarDate,isWeekEnd")
    trading_days = cal_dates[cal_dates['isWeekEnd']==1]["calendarDate"].tolist()
    trading_days = [day[0:4] + day[5:7] + day[8:] for day in trading_days]  # 更改日期的格式，将'2014-01-03'转化为'20140103'
    return trading_days

def get_index_cons(day="20140110", ticker="000300"):
    """获取当日指数的成份股，结果用字符串拼接起来方便后续的传参"""
    cons_id_df = DataAPI.IdxConsGet(secID=u"",ticker=ticker,intoDate=day,isNew=u"",field=u"consID",pandas="1")
    cons_id_ls = cons_id_df["consID"].tolist()
    return ",".join(cons_id_ls)

def plot_result(result_df, factor_name, ticker):
    """可视化最后的结果"""
    import matplotlib
    import matplotlib.pyplot as plt
    matplotlib.style.use('ggplot')
    result_df.plot(kind='bar', colors=['royalblue','crimson'],figsize=(16,6)).set_title(factor_name+" in "+ticker)
    plt.axhline(0, color='Gray')


def main(factor_name, ticker="000300", beging_day="20160101", end_day = "20160801"):
    """主函数，计算因子分位收益
       @factor_name 为因子名， 如"RSI","ADX"等
       @ticker 指数代码，如000300代表沪深300指数
       @beging_day 起始时间
       @end_day 结束时间
       """
    trading_days = get_weekend_trading_day_list(beging_day, end_day) 
    quantile_df = pd.DataFrame() #保存计算结果
    factor_api_field = "secID," + factor_name

    for i in xrange(len(trading_days)-1):
        sons_id_str = get_index_cons(trading_days[i],ticker)
        factor_df = DataAPI.MktStockFactorsOneDayGet(tradeDate=trading_days[i],secID=sons_id_str,ticker=u"",field=factor_api_field,pandas="1")  #获取每周最后一个交易日的因子值
        weekly_return = DataAPI.MktEquwAdjGet(secID=sons_id_str,beginDate=trading_days[i+1],endDate=trading_days[i+1],field=u"secID,return",pandas="1") # 获取相应股票未来一周的收益
        result = factor_df.merge(weekly_return,on='secID', how="inner")
        result[result["return"]==0] = None   #如果收益是0，非常有可能该股票一直处于停牌中，这里把种情况剔除
        result.dropna(inplace=True)
        
        result["10_quantile"] = pd.qcut(result[factor_name], 10, labels=False)  #按照因子的值平均划分为10份，并计算每一份的平均值
        lab_se = result.groupby("10_quantile")["return"].mean()
        quantile_df = pd.concat([quantile_df,lab_se], axis=1)
        
    quantile_df["mean"] = quantile_df.mean(axis=1)
    quantile_df.index = (quantile_df.index + 1) * 10
    return quantile_df["mean"]

def quantile_return(factor_name,ticker="000300", end_day="20160601"):
    previes_df = main(factor_name, ticker, "20140101", "20151231")
    this_year_df = main(factor_name, ticker, "20160101", end_day)
    result_df = pd.concat([previes_df,this_year_df], axis=1)
    result_df.columns = ["2014_2015", "2016"]
    result_df.to_csv("result.csv", index=False)
    # result_df = pd.read_csv("result.csv")
    plot_result(result_df, factor_name, ticker)
    



    
```

```python
quantile_return("RSI") 
```
