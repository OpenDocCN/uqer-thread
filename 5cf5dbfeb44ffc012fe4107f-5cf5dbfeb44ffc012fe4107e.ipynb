```python
badtype = ['投资者关系活动']
begindate = "20180417"
enddate = "20180620"
current_list = ['002117','002572']
checkdate = "20180416"
df = DataAPI.ReportContentGet(ticker=current_list,beginDate=begindate,endDate=enddate,field=['ticker','publishDate','reportType','title'],pandas="1")
df = df[~df.reportType.isin(badtype)].sort_values(by=['ticker','publishDate'],ascending=[1,1]).groupby(['ticker']).tail(1)
nobuy_list = df['ticker'].tolist()
print "nobuy_list:", nobuy_list
df_check = DataAPI.ReportContentGet(ticker=current_list,beginDate=checkdate,endDate=checkdate,field=['ticker','publishDate','reportType','title'],pandas="1")
df_check = df_check[~df_check.reportType.isin(badtype)].sort_values(by=['ticker','publishDate'],ascending=[1,1]).groupby(['ticker']).tail(1)
check_list = df_check['ticker'].tolist()
print "check_list:", check_list
prepare_list = list(set(current_list).intersection(check_list))
print "prepare_list:", prepare_list
prepared_list = list(set(prepare_list).difference(nobuy_list))
print "prepared_list:", prepared_list
if len(prepared_list) != 0:
    buy_universe = DataAPI.EquGet(ticker=prepared_list,equTypeCD=u"A",field=u"secID",pandas="1")['secID'].tolist()
    print "buy_universe:", buy_universe
else:
    buy_universe = []
    print "buy_universe:", buy_universe
```
