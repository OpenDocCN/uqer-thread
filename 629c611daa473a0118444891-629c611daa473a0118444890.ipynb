```python
from __future__ import division
import numpy as np
import pandas as pd
import matplotlib.pylab as plt
import seaborn as sns
import statsmodels.api as sm
import scipy.stats as st
import datetime
from sklearn.linear_model import LinearRegression
from matplotlib import rc
rc('mathtext', default='regular')
sns.set_style('white')
from matplotlib import dates
import time
from CAL.PyCAL import font    # CAL.PyCAL中包含font
from sklearn.decomposition import PCA

from functools import wraps
def wrapper_func_cache(func):
    @wraps(func)
    def wrapped(*args, **kwargs):
        if not hasattr(func, '__cache_result'):
            func.__cache_result = {}
        
        hash_key = [str(i) for i in args] + [str(j) for j in kwargs.iteritems()]
        cache_key = hash(tuple(hash_key))
        
        if cache_key not in func.__cache_result:
            func.__cache_result[cache_key] = func(*args, **kwargs)
            
        return func.__cache_result[cache_key]
    return wrapped

@wrapper_func_cache
def Monthly_const_trade_calendar(beginDate=u"20050101",endDate=u""):
    """交易日历，可以选择cache，只调用 api 一次
    """
    data = DataAPI.TradeCalGet(exchangeCD=u"XSHG,XSHE", beginDate=beginDate, endDate=endDate, field=['calendarDate', 'isMonthEnd'])
    data = data[data.isMonthEnd == 1].calendarDate.tolist()
    data = sorted(list(set(data))) 
    return data

# ST过滤
def st_remove(source_universe, st_date=None):
    """
    给定股票列表,去除其中在某日被标为ST的股票
    Args:
        source_universe (list of str): 需要进行筛选的股票列表
        st_date (datetime): 进行筛选的日期,默认为调用当天
    Returns:
        list: 去掉ST股票之后的股票列表

    Examples:
        >> universe = set_universe('A')
        >> universe_without_st = st_remove(universe)
    """
    st_date = st_date if st_date is not None else datetime.datetime.now().strftime('%Y%m%d')
    df_ST = DataAPI.SecSTGet(secID=source_universe, beginDate=st_date, endDate=st_date, field=['secID'])
    return [s for s in source_universe if s not in list(df_ST['secID'])]

# bug 函数封装问题
def winsorize_median(tmp):
    if isinstance(tmp, pd.Series):
        mc = sm.stats.stattools.medcouple(tmp)
        data = sorted(tmp)
        q1 = np.percentile(data, 25)
        q3 = np.percentile(data, 75)
        iqr = q3-q1
        if mc >= 0:
            l = q1-1.5*np.exp(-3.5*mc)*iqr
            u = q3+1.5*np.exp(4*mc)*iqr
        else:
            l = q1-1.5*np.exp(-4*mc)*iqr
            u = q3+1.5*np.exp(3.5*mc)*iqr
        tmp[tmp<l]=l
        tmp[tmp>u]=u
    return tmp   

# bug 函数封装问题
def standardize_zscore(tmp):
    if isinstance(tmp, pd.Series):
        mu = tmp.mean()
        sigma = tmp.std()
        tmp = (tmp - mu)/sigma
    return tmp   

# 获取申万一级行业分类成份股编码 
def Stock_Industry(point):
    # 申万一级指数编码
    industrys=DataAPI.IndustryGet(industryVersion=u"SW",industryVersionCD=u"",industryLevel=u"1",isNew=u"1",field=u"",pandas="1")
    indexes=list(industrys['indexSymbol'].astype(str))#行业对应的指数编码
    index_param=DataAPI.IdxGet(secID=u"",ticker=indexes,field=u"ticker,secShortName,publishDate",pandas="1")#获取指数上市时间即为对应行业开始时间
    #2014年前已存在的行业
    old_industry=index_param[index_param['publishDate']<'2014-01-01']
    old_indexes=list(old_industry['ticker'])
    old_industry_name=list(old_industry['secShortName'])
    #2014年后的所有行业
    new_industry=index_param
    new_indexes=list(new_industry['ticker'])
    new_industry_name=list(new_industry['secShortName'])
    # 申万一级指数分类函数
    SW_Old = old_indexes
    SW_New = new_indexes    
    if isinstance(point,str):
        point = pd.to_datetime(point)
    timepoint = pd.to_datetime('2014-01-01')
    if point <= timepoint:
        return SW_Old
    elif point > timepoint:
        return SW_New
    else:
        print 'wrong'

# 获取申万一级行业分类成份股       
def Industry_Constituent_Stocks(industry, timepoint):
    tmp = DataAPI.IdxConsGet(ticker=industry, intoDate=timepoint, isNew=u"",field=['consID'],pandas="1")
    return sorted(list(tmp.consID.values))
```

```python
all_date=pd.DataFrame()
trade_calendar = Monthly_const_trade_calendar(beginDate=u"20140601",endDate=u"20160601")

for date in trade_calendar[:-1]:
    
    # print date
    # 获取当日股票池
    univ = set_universe('A',date)
    
    univ = st_remove(univ, date)
    # 获取横截面股票收益率
    date_next = trade_calendar[trade_calendar.index(date)+1]
    Price_front = DataAPI.MktEqudGet(tradeDate=date, secID=univ,isOpen="1",
                            field=['tradeDate','secID','closePrice','negMarketValue'],
                                     pandas="1")
    Price_front['close']=Price_front['closePrice']
    Price_behind = DataAPI.MktEqudGet(tradeDate=date_next, secID=univ, 
                                      field=['secID','closePrice'],pandas="1")
    
    R_T = pd.merge(Price_front.loc[:,['tradeDate','secID','close','negMarketValue']], Price_behind.loc[:,['secID','closePrice']], on='secID')
   
    R_T['fac_MV'] = standardize_zscore(winsorize_median(R_T.negMarketValue))
    R_T['return'] = (R_T.closePrice-R_T.close)/R_T.close
    SW_Industry_List = Stock_Industry(date)
   
    
    # 获取横截面因子载荷
    factor_loadings = DataAPI.MktStockFactorsOneDayGet(tradeDate=date, secID=univ, field='secID,PB,PE,ROE,VOL60,DHILO,DVRAT,HSIGMA',pandas="1")
    factor_loadings.dropna(inplace=True)
    for fac in factor_loadings.columns[1:]:
        factor_loadings[fac] = standardize_zscore(winsorize_median(factor_loadings[fac]))
    #factor_loadings[fac].plot()
        factor_loadings.set_index('secID',inplace=True, drop=False)
        Linear_Regression = pd.DataFrame()
        for i in SW_Industry_List:
            i_Constituent_Stocks = Industry_Constituent_Stocks(i, date)
            i_Constituent_Stocks = list(set(i_Constituent_Stocks).intersection(set(univ)))
            try:
                tmp = factor_loadings.loc[i_Constituent_Stocks]
                tmp.dropna(inplace=True)
                tmp[i]=1.0
            except:
                print i
            Linear_Regression = Linear_Regression.append(tmp)
        Linear_Regression.fillna(0.0, inplace=True)
        Linear_Regression = pd.merge(Linear_Regression, R_T.loc[:,['tradeDate','secID','return','fac_MV']], on='secID')
        Linear_Regression.set_index('secID', inplace=True)
        
        X = Linear_Regression.iloc[:,:28]
        
        X['fac_MV'] = Linear_Regression.fac_MV
        
        y = Linear_Regression.loc[:,fac]
        
        # OLS回归
        est=sm.OLS(y,X)
        result = est.fit()
        Linear_Regression.loc[:,fac] = result.resid
        
    all_date=pd.concat([all_date,Linear_Regression[['tradeDate','PB','PE','ROE','VOL60','DHILO',
                                                    'DVRAT','HSIGMA','return']]])

X,y=all_date[['PB','PE','ROE','VOL60','DHILO','DVRAT','HSIGMA']].values,all_date['return'].values
pca=PCA(n_components=5)
X_=pca.fit_transform(X)
lr=LinearRegression()
lr.fit(X_,y)
coef_=lr.coef_
coef_
```

```python
fac = 'ROE'

# 可编辑部分与 strategy 模式一样，其余部分按本例代码编写即可

# -----------回测参数部分开始，可编辑------------
start = '2017-11-01'                       # 回测起始时间
end = '2018-01-01'                         # 回测结束时间
benchmark = 'HS300'                        # 策略参考标准
universe = DynamicUniverse('HS300')  # 证券池，支持股票和基金
capital_base = 1e9   
freq = 'd'
refresh_rate = Monthly(-1)  
# ---------------回测参数部分结束----------------
# 把回测参数封装到 SimulationParameters 中，供 quick_backtest 使用
sim_params = quartz.SimulationParameters(start, end, benchmark, universe, capital_base)
# 获取回测行情数据
idxmap, data = quartz.get_daily_data(sim_params)
# 运行结果
results = {}
# 调整参数(选取股票的Q因子五分位数)，进行快速回测
for quantile_five in range(0, 5):
    
    # ---------------策略逻辑部分----------------
    refresh_rate = Monthly(-1)  
    commission = Commission(0.0002, 0.0002)
    
    def initialize(account):                   # 初始化虚拟账户状态
        pass
    
    def handle_data(account):               

        Operation_Date = account.current_date.strftime('%Y%m%d') 

        # 过滤ST股票
        univ = account.universe
        #print len(univ)

        # 判断行业分类
        Industrys = Stock_Industry(account.current_date)

        # 获取行业划分成份股(将沪深300里面的股票按照行业进行分类{行业1：[],行业2：[]})
        Constituent_Stocks = {}
        for s in Industrys:
            tmp = Industry_Constituent_Stocks(s, Operation_Date)
            tmp = list(set(tmp).intersection(set(univ)))
            Constituent_Stocks[s] = tmp

        # 获取因子数据
        factor_ = DataAPI.MktStockFactorsOneDayGet(tradeDate=account.previous_date.strftime('%Y%m%d'),
                                                   secID=univ, field=['secID']+[fac], pandas="1")
        
        factor_.dropna(inplace=True)
        factor_.set_index('secID', inplace=True)
        
        # 按行业对行业内股票进行排序筛选
        holding_list = []
        holding = {}
        for s in Industrys:
            tmp = Constituent_Stocks[s]
            tmp = factor_.loc[tmp,:]
            tmp.sort_values(fac, inplace=True)
            q_min = tmp.quantile(quantile_five*0.2)[fac]
            q_max = tmp.quantile((quantile_five+1)*0.2)[fac]
            industry_univ = tmp[(tmp[fac]>=q_min)&(tmp[fac]<=q_max)]
            industry_univ = industry_univ.index.values.tolist()
            holding[s] = industry_univ
            holding_list = holding_list + industry_univ

        # 仓位操作
        # 不在股票池中的，清仓
        for stk in account.security_position:
            if stk not in holding_list:
                account.order_to(stk, 0)
        # 在目标股票池中的，等权买入
        for stk in holding_list:
            order_pct_to(stk, 1./len(holding_list))
# ---------------策略逻辑部分结束----------------

    # 把回测逻辑封装到 TradingStrategy 中，供 quick_backtest 使用
    strategy = quartz.TradingStrategy(initialize, handle_data)
    # 回测部分
    bt, acct = quartz.quick_backtest(sim_params, strategy, idxmap, data, refresh_rate=refresh_rate, commission=commission)

 
    # 对于回测的结果，可以通过 perf_parse 函数计算风险指标
    perf = quartz.perf_parse(bt, acct)

    # 保存运行结果
    tmp = {}
    tmp['bt'] = bt
    tmp['annualized_return'] = perf['annualized_return']
    tmp['volatility'] = perf['volatility']
    tmp['max_drawdown'] = perf['max_drawdown']
    tmp['alpha'] = perf['alpha']
    tmp['beta'] = perf['beta']
    tmp['sharpe'] = perf['sharpe']
    tmp['information_ratio'] = perf['information_ratio']
    
    results[quantile_five] = tmp
    print str(quantile_five)
print 'done'   
```

```python
import matplotlib.pyplot as plt
from matplotlib import rc
rc('mathtext', default='regular')
import seaborn as sns
sns.set_style('white')
from matplotlib import dates
import time
import scipy.stats as st
from CAL.PyCAL import font    # CAL.PyCAL中包含font


fig = plt.figure(figsize=(14,9))
fig.set_tight_layout(True)
ax1 = fig.add_subplot(211)
ax2 = fig.add_subplot(212)
ax1.grid()
ax2.grid()

for qt in results:
    bt = results[qt]['bt']

    data = bt[[u'tradeDate',u'portfolio_value',u'benchmark_return']]
    data['portfolio_return'] = data.portfolio_value/data.portfolio_value.shift(1) - 1.0   # 总头寸每日回报率
    data['portfolio_return'].ix[0] = data['portfolio_value'].ix[0]/	1e9 - 1.0
    data['excess_return'] = data.portfolio_return - data.benchmark_return                 # 总头寸每日超额回报率
    data['excess'] = data.excess_return + 1.0
    data['excess'] = data.excess.cumprod()                # 总头寸对冲指数后的净值序列
    data['portfolio'] = data.portfolio_return + 1.0     
    data['portfolio'] = data.portfolio.cumprod()          # 总头寸不对冲时的净值序列
    data['benchmark'] = data.benchmark_return + 1.0
    data['benchmark'] = data.benchmark.cumprod()          # benchmark的净值序列
    results[qt]['hedged_max_drawdown'] = max([1 - v/max(1, max(data['excess'][:i+1])) for i,v in enumerate(data['excess'])])  # 对冲后净值最大回撤
    results[qt]['hedged_volatility'] = np.std(data['excess_return'])*np.sqrt(252)
    results[qt]['hedged_annualized_return'] = (data['excess'].values[-1])**(252.0/len(data['excess'])) - 1.0
    # data[['portfolio','benchmark','excess']].plot(figsize=(12,8))
    # ax.plot(data[['portfolio','benchmark','excess']], label=str(qt))
    ax1.plot(data['tradeDate'], data[['portfolio']], label=str(qt))
    ax2.plot(data['tradeDate'], data[['excess']], label=str(qt))
    

ax1.legend(loc=0)
ax2.legend(loc=0)
ax1.set_ylabel(u"净值", fontproperties=font, fontsize=16)
ax2.set_ylabel(u"对冲净值", fontproperties=font, fontsize=16)
```
