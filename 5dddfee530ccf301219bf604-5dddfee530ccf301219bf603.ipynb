# 波动率与换手率构建牛熊指标

```python
import math
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# pd.set_option("display.max_rows",200)
# pd.set_option("display.max_columns",300)
```

# 1、波动率的研究

```python
# 波动率的研究
#取出深300指数,开盘价，最低价、最高价和收盘价
startdate = '20090301'
enddate = '20191120'
index_samp = '000300'
window_list = [30,60,120,200,250]
colors = ['gold','b','orange','green','purple']
# index_factor = ['tradeDate','openIndex','lowestIndex','highestIndex','closeIndex','CHGPct']
index_factor = ['tradeDate','closeIndex','CHGPct']
price_data = DataAPI.MktIdxdGet(ticker=index_samp,beginDate=startdate,endDate=enddate,exchangeCD=u"XSHG",field=index_factor,pandas="1")
price_data=price_data.set_index('tradeDate')
# 波动率指标的计算主要基于股票指数的收益率，是过去历史n个交易日收益率的标准差。
for window in window_list:
    price_data['CHGPct_rolling_%s'%str(window)] = price_data['CHGPct'].rolling(window=window,center=False).std()
price_data = price_data.dropna()
print (price_data.shape)
print (price_data.head().to_html())
print (price_data.tail().to_html())

plt.figure(figsize=(20,6), dpi=80)
for window in window_list:
    price_data['CHGPct_rolling_%s'%str(window)].plot(label='CHGPct_rolling_%s'%str(window),c=colors[window_list.index(window)],fontsize=15)
plt.xticks(rotation=20)
plt.legend(loc='best',fontsize=15)
plt.grid(False) 
plt.savefig('.//volatility_turnoverRate/CHGPct_rolling_%s.jpg'%index_samp)
plt.show();
```

# 2、换手率的研究

```python
# 换手率的研究
index_factor1 =  ['tradeDate','PEType','turnoverRate']
# turnoverRate	float	换手率，成分成交额之和/成分流通市值之和。
index_info = DataAPI.MktIdxdEvalGet(ticker=index_samp,beginDate=startdate,endDate=enddate,field=index_factor1,pandas="1")
index_info.drop_duplicates('tradeDate',keep='first',inplace=True)
index_info=index_info.set_index('tradeDate')
for window in window_list:
    index_info['turnoverRate_%s'%str(window)] = index_info['turnoverRate'].rolling(window=window,center=False).mean()
index_info = index_info.dropna()
print (index_info.shape)
print (index_info.head().to_html())
print (index_info.tail().to_html())

plt.figure(figsize=(20,6), dpi=80)
# colors = ['gold','b','orange','green','purple']
for window in window_list:
    index_info['turnoverRate_%s'%str(window)].plot(label='turnoverRate_%s'%str(window),c=colors[window_list.index(window)],fontsize=15)
plt.xticks(rotation=20)
plt.legend(loc='best',fontsize=15)
plt.grid(False) 
plt.savefig('.//volatility_turnoverRate/turnoverRate_%s.jpg'%index_samp)
plt.show();
```

# 3、波动率/换手率构建牛熊指标

```python
factor_merge = pd.merge(price_data,index_info,left_index=True,right_index=True)
for window in window_list:
    factor_merge['CHGPct_rolling/turnoverRate_%s'%str(window)] = factor_merge['CHGPct_rolling_%s'%str(window)]/factor_merge['turnoverRate_%s'%str(window)]
factor_merge = factor_merge.dropna()
print (factor_merge.shape)
print (factor_merge.head().to_html())
print (factor_merge.tail().to_html())

plt.figure(figsize=(20,6), dpi=80)
# colors = ['gold','b','orange','green','purple']
for window in window_list:
    factor_merge['CHGPct_rolling/turnoverRate_%s'%str(window)].plot(label='CHGPct_rolling/turnoverRate_%s'%str(window),c=colors[window_list.index(window)],fontsize=15)
plt.xticks(rotation=20)
plt.legend(loc='best',fontsize=15)
plt.grid(False) 
plt.savefig('.//volatility_turnoverRate/CHGPct_rolling_turnoverRate_%s.jpg'%index_samp)
plt.show();
```

# 4、牛熊指标双均线与指数走势

```python
plt.figure(figsize=(20,6), dpi=80)
# colors = ['b','orange','green','purple']
window = 30
# factor_merge['CHGPct_rolling/turnoverRate_%s'%window].rolling(window=20,min_periods=1).mean().plot(label='CHGPct_rolling/turnoverRate_%s_20'%window,c='b',fontsize=15)
# factor_merge['CHGPct_rolling/turnoverRate_%s'%window].rolling(window=60,min_periods=1).mean().plot(label='CHGPct_rolling/turnoverRate_%s_60'%window,c='orange',fontsize=15)
factor_merge['CHGPct_rolling/turnoverRate_%s_20'%window] = factor_merge['CHGPct_rolling/turnoverRate_%s'%window].rolling(window=20,min_periods=1).mean()
factor_merge['CHGPct_rolling/turnoverRate_%s_60'%window] = factor_merge['CHGPct_rolling/turnoverRate_%s'%window].rolling(window=60,min_periods=1).mean()
factor_merge['closeIndex_20'] = factor_merge['closeIndex'].rolling(window=20,min_periods=1).mean()
factor_merge['closeIndex_60'] = factor_merge['closeIndex'].rolling(window=60,min_periods=1).mean()
factor_merge['CHGPct_rolling/turnoverRate_%s_20'%window].plot(label='CHGPct_rolling/turnoverRate_%s_20'%window,c='b',fontsize=15)
factor_merge['CHGPct_rolling/turnoverRate_%s_60'%window].plot(label='CHGPct_rolling/turnoverRate_%s_60'%window,c='orange',fontsize=15)
plt.legend(loc='best',fontsize=15)
plt.twinx()
# factor_merge['closeIndex_20'].plot(label='closeIndex_20',c='brown',fontsize=15)
# factor_merge['closeIndex_60'].plot(label='closeIndex_60',c='gold',fontsize=15)
factor_merge['closeIndex'].plot(label='closeIndex',c='r',fontsize=15)
plt.xticks(rotation=20)
plt.legend(loc='upper left',fontsize='20')
plt.grid(False) 
# plt.savefig('.//volatility_turnoverRate/CHGPct_rolling_turnoverRate_%s.jpg'%index_samp)
plt.show();
```

# 5、牛熊指标双均线策略

```python
detail_df = factor_merge[60:]
factors = ['closeIndex','CHGPct','CHGPct_rolling/turnoverRate_%s_20'%window,'CHGPct_rolling/turnoverRate_%s_60'%window,'closeIndex_20','closeIndex_60']
detail_df = detail_df[factors]
detail_df['CHGPct_rolling/turnoverRate_%s_20_shift'%window] = detail_df['CHGPct_rolling/turnoverRate_%s_20'%window].shift(1)
detail_df['CHGPct_rolling/turnoverRate_%s_60_shift'%window] = detail_df['CHGPct_rolling/turnoverRate_%s_60'%window].shift(1)
detail_df['closeIndex_20_shift'] = detail_df['closeIndex_20'].shift(1)
detail_df['closeIndex_60_shift'] = detail_df['closeIndex_60'].shift(1)
detail_df.dropna(inplace=True)

detail_df['signal_index'] = 0
detail_df['signal_index'][detail_df['closeIndex_60_shift']<detail_df['closeIndex_20_shift']] = 1
detail_df['return_index'] = 0
detail_df['return_index'][detail_df['signal_index'] == 1] = detail_df['CHGPct'][detail_df['signal_index'] == 1] 
detail_df['excess_return_index'] = detail_df['return_index'] - detail_df['CHGPct']
detail_df['value_index'] = (detail_df['return_index']+1).cumprod()

detail_df['signal'] = 0
detail_df['signal'][detail_df['CHGPct_rolling/turnoverRate_%s_20_shift'%window]<detail_df['CHGPct_rolling/turnoverRate_%s_60_shift'%window]] = 1
detail_df['return'] = 0
detail_df['return'][detail_df['signal'] == 1] = detail_df['CHGPct'][detail_df['signal'] == 1] 
detail_df['excess_return'] = detail_df['return'] - detail_df['CHGPct']
detail_df['value'] = (detail_df['return']+1).cumprod()
detail_df['index_cumprod_value'] = (detail_df['CHGPct']+1).cumprod()
print (detail_df.shape)
print (detail_df.head().to_html())
print (detail_df.tail().to_html())
```

```python
# --evaluation
# 计算策略的年化超额收益：所有收益的均值乘以250。
ann_excess_return = np.mean(detail_df['excess_return'])*250
# 计算策略的年化超额收益波动率
ann_excess_vol = np.std(detail_df['excess_return'])*np.sqrt(250)
# 计算策略的信息比率IR：超额收益/超额收益波动率。用于评估策略好坏，含义是承担单位主动风险所获得的超额收益，IR 越大则策略越好，反之越差。
info_ratio = ann_excess_return/ann_excess_vol

# --print out
print ('annual excess return = %.3f'%ann_excess_return)
print ('annual excess volatility = %.3f'%ann_excess_vol)
print ('information ratio = %.3f'%info_ratio)
```
