### MACD为例
以MACD为例:
百度到的买卖原则为：
1.DIF、DEA均为正，DIF向上突破DEA，买入信号参考。
2.DIF、DEA均为负，DIF向下跌破DEA，卖出信号参考。
3.DIF线与K线发生背离，行情可能出现反转信号。
4.DIF、DEA的值从正数变成负数，或者从负数变成正数并不是交易信号，因为它们落后于市场。
有点复杂，简化：dif向上突破dea，买入。下突破dea，卖出
如何将这些信息建模从而传递给算法呢（算法是不认识”突破“的）？

观察macd曲线可以看到，
除了交叉点可能是拐点外，dea，dif的相对位置也比较重要，基本上可以大致认为牛熊状态
所以macd中我们只使用dea，dif，基于此创建信息
1. dif上突破dea，1。下突破-1，其他0
2. dif高于dea，1。dif低于dea-1，粘滞0（粘滞的识别？）

这样的话上面的两个重要特征都可以传递给算法了。
但技术的交叉可能一天完成，交叉次日就消失，过于严格会导致信息低频，故新增规则
3. 突破，交叉信号发出后，会持续n天(n=3，比如T日交叉，T+1，T+2虽然没交叉，但是也认为交叉了)

同理，其他指标kdj，bolling等都需要转为1，-1，0信号，
这样的话如果有模式”macd金叉且kdj金叉时x日后收益为正的可能较大“时，算法便更容易捕捉到

## 第二步：数据准备
技术因子：MACD,KDJ,RSI,DMI为例
算法：随机森林

```python
import time
import pandas as pd

tec_factor_list=['MACD_STATUS','MACD_SIGNAL','KDJ_STATUS','KDJ_SIGNAL','RSI_SIGNAL','DMI_STATUS','DMI_SIGNAL']

def getTecFacAll(universe, begin_date, end_date):
    # 计算各股票历史区间回报率，过去window天的收益率
    print ' TecFacAll will be calculated for ' + str(len(universe)) + ' stocks:'
    count = 0
    N=50
    secs_time = 0
    start_time = time.time()
    # 构建日期列表
    data=DataAPI.TradeCalGet(exchangeCD=u"XSHG",beginDate=begin_date,endDate=end_date,field=['calendarDate','isOpen'],pandas="1")
    data=data[data['isOpen']==1]
    date_list = data['calendarDate'].values.tolist()
    all_data_factor=['secID','tradeDate','returnRate5','returnRate10']+tec_factor_list
    all_data=pd.DataFrame(columns=all_data_factor)
    for stk in universe:
        try:
            #MACD,status-dif>dea则1其他-1，signal-dif金叉dea则1死叉-1其他0
            sec_data=DataAPI.MktStockFactorsDateRangeGet(secID=stk,ticker='',\
                                                         beginDate=begin_date,endDate=end_date,\
                                                         field=u"secID,tradeDate,DIFF,DEA,KDJ_K,KDJ_D,RSI,plusDI,minusDI"
                                                         ,pandas="1")
            sec_data=sec_data.sort(columns=['tradeDate'])
            sec_data['tradeDateIndex']=sec_data['tradeDate']
            sec_data=sec_data.set_index('tradeDateIndex')

            sec_data['MACD_STATUS']=(sec_data['DIFF']>sec_data['DEA']).apply(lambda x:1 if x>0 else -1)
            sec_data['MACD_SIGNAL']=sec_data['MACD_STATUS'].diff().apply(lambda x:1 if x>0.01 else -1 if x<-0.01 else 0)
            #KDJ,k>d则status=1其他-1,singnal-k金叉d则1死叉-1其他0
            sec_data['KDJ_STATUS']=(sec_data['KDJ_K']>sec_data['KDJ_D']).apply(lambda x:1 if x>0 else -1)
            sec_data['KDJ_SIGNAL']=sec_data['KDJ_STATUS'].diff().apply(lambda x:1 if x>0.01 else -1 if x<-0.01 else 0)
            sec_data['RSI_SIGNAL']=0
            #RSI,rsi向上突破501向下-1，20以下3,80以上4
            for i in xrange(1,len(sec_data)):
                if sec_data['RSI'][i]>50 and sec_data['RSI'][i-1]<=50:
                    sec_data['RSI_SIGNAL'][i]=1
                elif sec_data['RSI'][i]<=50 and sec_data['RSI'][i-1]>50:
                    sec_data['RSI_SIGNAL'][i]=-1
                elif sec_data['RSI'][i]<20:
                    sec_data['RSI_SIGNAL'][i]=3
                elif sec_data['RSI'][i]>80:
                    sec_data['RSI_SIGNAL'][i]=4
                else:
                    sec_data['RSI_SIGNAL'][i]=0

            #DMI,+DI>-DI则status=1其他-1,singnal-+DI金叉-DI则1死叉-1其他0
            sec_data['DMI_STATUS']=(sec_data['plusDI']>sec_data['minusDI']).apply(lambda x:1 if x>0 else -1)
            sec_data['DMI_SIGNAL']=sec_data['DMI_STATUS'].diff().apply(lambda x:1 if x>0.01 else -1 if x<-0.01 else 0)

            #计算5日,10日后收益
            price_data=DataAPI.MktEqudAdjGet(tradeDate=u"",secID=stk,\
                                             ticker=u"",isOpen="",beginDate=begin_date,endDate=end_date,\
                                             field=u"secID,tradeDate,closePrice,preClosePrice",pandas="1")
            price_data=price_data.sort(columns=['tradeDate']).set_index('tradeDate')
            sec_data['returnRate5']=(price_data['closePrice'].shift(-5)\
                                     .fillna(method='pad')/price_data['closePrice']-1)
            sec_data['returnRate10']=(price_data['closePrice'].shift(-10)\
                                      .fillna(method='pad')/price_data['closePrice']-1)
            all_data=all_data.append(sec_data[all_data_factor])

            # 打印进度部分
            count += 1
            if count > 0 and count % N == 0:
                finish_time = time.time()
                print count,
                print '  ' + str(finish_time-start_time-secs_time) + ' seconds elapsed.'
                secs_time = (finish_time-start_time)
        except Exception,e:
            print e
    
    all_data.to_csv('zall_data.csv')
    return 
getTecFacAll(set_universe('HS300'), '20150101', '20170101')
```

## 第三步:预测准确率统计

```python
from sklearn.ensemble import RandomForestClassifier
from CAL.PyCAL import *
import pandas as pd
from datetime import *
import CAL.PyCAL as pycal
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import tree
import time
from sklearn.naive_bayes import GaussianNB


def extendSignal(value_list=[],num=5):
    #将信号1和-1向后延迟num单位
    p_value=0#当前值
    p_value_count=0#当前值的赋值次数，每次赋值后-1
    for i in xrange(0,len(value_list)):
        if value_list[i]!=0:
            p_value=value_list[i]
            p_value_count=num
        elif p_value!=0 and p_value_count>0:
            value_list[i]=p_value
            p_value_count-=1
    return value_list

train_factors=['LCAP', 'LFLO', 'NPToTOR', 'OperatingProfitGrowRate', 'TotalAssetGrowRate', 'DHILO', 'DEGM', 'Skewness', 'DAREC', 'GREC']
tec_factor_list=['MACD_STATUS','MACD_SIGNAL','KDJ_STATUS','KDJ_SIGNAL','RSI_SIGNAL','DMI_STATUS','DMI_SIGNAL']

all_data = pd.read_csv('zall_data.csv')

#信号拓展3天
all_data['MACD_SIGNAL']=extendSignal(all_data['MACD_SIGNAL'].values.tolist(),3)
all_data['KDJ_SIGNAL']=extendSignal(all_data['KDJ_SIGNAL'].values.tolist(),3)
all_data['RSI_SIGNAL']=extendSignal(all_data['RSI_SIGNAL'].values.tolist(),3)
all_data['DMI_SIGNAL']=extendSignal(all_data['DMI_SIGNAL'].values.tolist(),3)

#收益离散化
all_data['returnRate5']=all_data['returnRate5'].apply(lambda x:1 if x>0.015 else -1 )

#进度打印
count=0
N=50
secs_time=0
def getPredict(predict_date,window=20,stocklist=[],filename='',train_factor=[]):
    #结束日期为predict-1工作日
    cal = Calendar('China.SSE')
    pdate=date(int(predict_date[0:4]),int(predict_date[5:7]),int(predict_date[9:]))
    yest_date = cal.advanceDate(pdate, '-1B', BizDayConvention.Following).strftime('%Y-%m-%d')
    train_start_date=cal.advanceDate(pdate,'-'+str(1+5+window)+'B',BizDayConvention.Following).strftime('%Y-%m-%d')
    train_end_date=cal.advanceDate(pdate,'-'+str(1+5)+'B',BizDayConvention.Following).strftime('%Y-%m-%d')
    
    #获取训练数据，日期过滤和代码过滤
    #all_data=pd.read_csv(filename)
    train_data_df=all_data[all_data['tradeDate']>=train_start_date]
    train_data_df=train_data_df[train_data_df['tradeDate']<=train_end_date]
    
 
    train_data=train_data_df[train_factor].values.tolist()
    train_label=train_data_df['returnRate5'].values.tolist()
    #获取预测数据
    predict_data_df=all_data[all_data['tradeDate']==yest_date][all_data['secID'].isin(stocklist)].dropna()
    predict_data=predict_data_df[train_factor].values.tolist()
   
    # 创建并训练随机森林,根据上一个交易日的因子预测涨跌,返回预测涨幅最大的前20支股票
    
    rf = RandomForestClassifier(n_estimators = 1000)  #1000个决策树
    rf.fit(train_data, train_label)
    #print rf.score(train_data, train_label)
 
    predict_data_df['predict']=rf.predict(predict_data)    
    #print predict_data_df['predict']
    predict_data_df=predict_data_df.set_index('secID')
    predict_data_df=predict_data_df.sort(columns='predict',ascending=False)
    return predict_data_df['predict'].to_dict()



#stock_data=getStockFactor('2017-02-05','2017-03-20',secID='600006.XSHG')
'''print getPredict('2017-04-25',window=50,\
                 stocklist=['600006.XSHG','000021.XSHE'],\
                 filename='zall_data.csv',train_factor=tec_factor_list)'''

#工作日差相关参数初始化
calSSE=pycal.Calendar('China.SSE')
dateList=calSSE.bizDatesList(pycal.Date.strptime('2015-01-01','%Y-%m-%d'),\
                             pycal.Date.strptime('2017-01-01','%Y-%m-%d'))    
dateStrList=[]#yyyymmdd格式日期串
for datetmp in dateList:
    dateStrList.append(datetmp.strftime('%Y-%m-%d'))
universe=set_universe('HS300')
right_rate={}
predict_Series=pd.DataFrame(index=dateStrList,columns=universe)
start_time=time.time()
time.time()
for dateStr in dateStrList:
    try:
        p_result=getPredict(dateStr,window=10,\
                     stocklist=universe,\
                     filename='zall_data.csv',train_factor=tec_factor_list)
        #统计正确率,预测为1结果也为1的认为正确，预测为0或者-1的不考虑
        
        predict_buy=[x for x,y in p_result.items() if y==1 ]
        
        real_result=all_data[all_data['tradeDate']==dateStr][all_data['secID'].isin(predict_buy)]
        
        right_rate[dateStr]=real_result['returnRate5'].value_counts()[1]*1.0/len(predict_buy)
        predict_Series.loc[dateStr]=pd.Series(p_result)
        # 打印进度部分
        count += 1
        if count > 0 and count % N == 0:
            finish_time = time.time()
            print count,' use time:' + str(finish_time-start_time-secs_time) + ' seconds elapsed.'
            secs_time = (finish_time-start_time)
    except Exception,e :
        pass#print e
print right_rate
pd.Series(right_rate).plot()
predict_Series.to_csv('zpredict_series.csv')
```
