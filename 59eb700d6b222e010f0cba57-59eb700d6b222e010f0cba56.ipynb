# 焦炭焦煤价差研究

```python
#获取数据
import numpy as np
import pandas as pd
import matplotlib.pylab as plt
from scipy import stats
#期货主力、连续合约日行情
field = ['tradeDate', 'ticker', 'contractObject', 'openPrice', 'highestPrice', 'lowestPrice', 'closePrice', 'settlePrice', 'mainCon']
jmData = DataAPI.MktMFutdGet(tradeDate=u"",mainCon=u"",contractMark=u"",contractObject=u"JM",startDate=u"",endDate=u"",field=field,pandas="1")
jmData = jmData.ix[jmData['mainCon'] == 1]
jmData.drop(['contractObject', 'mainCon'], axis = 1, inplace = True)
jmData.index = range(len(jmData))

jtData = DataAPI.MktMFutdGet(tradeDate=u"",mainCon=u"",contractMark=u"",contractObject=u"J",startDate=u"",endDate=u"",field=field,pandas="1")
jtData = jtData.ix[jtData['mainCon'] == 1]
jtData.index = range(len(jtData))
idx = jtData.ix[jtData['tradeDate'] == '2013-03-22'].index.values[0]
jtData = jtData[idx:]
jtData.drop(['contractObject', 'mainCon'], axis = 1, inplace = True)
jtData.index = range(len(jtData))
jtData
#合并焦煤和焦炭数据
jmjtData = jmData.merge(jtData, left_on = ['tradeDate'], right_on = ['tradeDate'], how = 'inner', suffixes=('_jm', '_jt'))
#计算焦炭和焦煤的价差
jmjtData['diff'] = jmjtData['closePrice_jt'] - jmjtData['closePrice_jm']
#jmjtData['rate'] = jmjtData['closePrice_jt'] / jmjtData['closePrice_jm'] #计算比值
```

```python
#绘制焦煤焦炭价格和其价差的图
plt.figure(figsize = (14,7))
plt.grid()
plt.plot(pd.to_datetime(jmjtData['tradeDate']), jmjtData['closePrice_jm'], label = 'JT', linewidth=2)
plt.plot(pd.to_datetime(jmjtData['tradeDate']), jmjtData['closePrice_jt'], label = 'JM', linewidth=2)
plt.plot(pd.to_datetime(jmjtData['tradeDate']), jmjtData['diff'], '--', label = 'diff', linewidth=2)
plt.xlabel('date')
plt.ylabel('price')
plt.title('price of JM & JT & diff')
plt.legend(loc = 'best')
plt.show()
```

```python
#焦煤焦炭价格的相关关系分析
print '焦煤和焦炭收盘价的相关系数为：', (np.corrcoef(jmjtData['closePrice_jm'], jmjtData['closePrice_jt'])[0,1]) * 100 #以收盘价为准
#np.corrcoef(jmjtData['settlePrice_jm'], jmjtData['settlePrice_jt']) #以结算价为准
print '价差与焦炭收盘价的相关系数为：', (np.corrcoef(jmjtData['diff'], jmjtData['closePrice_jt'])[0,1]) * 100 #以收盘价为准
print '价差与焦煤收盘价的相关系数为：', (np.corrcoef(jmjtData['diff'], jmjtData['closePrice_jm'])[0,1]) * 100 #以收盘价为准

#进行一元线性回归拟合
slope, intercept, _, _ ,_ = stats.linregress(jmjtData['closePrice_jm'], jmjtData['closePrice_jt']) #得到焦炭和焦煤价格的拟合方程：JT = -158.8621 + 1.5722 * JM 

#绘制焦煤焦炭收盘价的散点图和拟合图
plt.figure(figsize = (14,7))
plt.grid()
plt.plot(jmjtData['closePrice_jm'], jmjtData['closePrice_jt'], 'o', label='original data')
plt.plot(jmjtData['closePrice_jm'], intercept + slope*jmjtData['closePrice_jm'], 'r', linewidth=2, label='fitted line')
plt.legend(loc = 'best')
plt.xlabel('JT')
plt.ylabel('JM')
plt.title('scatter of JM & JT closePrice')
plt.show()
```

```python
#焦炭焦煤价差的描述性统计
print 'mean: 均值'
print 'median: 中位数'
print 'mode: 众数'
print 'max: 最大值'
print 'min: 最小值'
print 'ptp: 极差'
print 'std: 标准差'
print 'cv: 变异系数'
description = pd.DataFrame(index=['value'],columns=['mean','median','mode','max', 'min', 'ptp', 'std', 'cv'])
description['mean'] = np.mean(jmjtData['diff'])
description['median'] = np.median(jmjtData['diff'])
description['mode'] = stats.mode(jmjtData['diff'])[0][0]
description['max'] = np.max(jmjtData['diff'])
description['min'] = np.min(jmjtData['diff'])
description['ptp'] = np.ptp(jmjtData['diff'])
description['std'] = np.std(jmjtData['diff'])
description['cv'] = description['mean'] / description['std']
description
```

```python
#价差的偏离程度分析
qlow = np.percentile(jmjtData['diff'], 25)
print '价差的下四分位数为：', qlow
median = np.percentile(jmjtData['diff'], 50)
print '价差的中位数为：', median
qup = np.percentile(jmjtData['diff'], 75)
print '价差的上四分位数为：', qup
IQR = np.percentile(jmjtData['diff'], 75) - np.percentile(jmjtData['diff'], 25)
print '价差的四分位距为：', qup - 1.5*IQR
print '箱线图的最上线为：', qlow + 1.5*IQR
print '箱线图的最下线为：', IQR

#绘制箱线图
plt.figure(figsize = (14,7))
plt.grid()
plt.boxplot(np.array(jmjtData['diff']), labels = ['DIFF'])
plt.title('boxplot of diff')
plt.show()
```

```python
几点结论：
1、双焦下跌行情中，双焦价差缩小，双焦上涨行情中，双焦价差增大。言外之意就是，下跌行情多焦煤空焦炭，上涨行情多焦炭空焦煤。
2、焦煤和焦炭收盘价的相关系数竟然高达97.62%，然而他们的价差和焦炭的相关系数也达到了94.63%，和焦煤的相关系数只有85.38%，比焦炭少了接近10个百分点。
3、历史最大价差竟然达到1005元之大，最小只有71元，而价差均值则在359元。
4、历史价差绝大部分都分布在，10%（166元）以上，80%（468元）以下。
5、价差数据严重正偏，且价差分布在600元以下的概率达到了90%以上！
必须说明的是以上结论都是基于历史数据得出，而且历史数据只有4年多，而双焦产业链价格周期可能都有几年之久，结论仅做参考之用！
```
