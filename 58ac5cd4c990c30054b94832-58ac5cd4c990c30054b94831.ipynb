# 一个风险收益的小策略

```python
import numpy as np
import pandas as pd
import datetime
import time
import matplotlib.pyplot as plt
```

```python
##股票选取
def chs_factor(ticker = '000005',begin = '20010101',end = '20170201',tradeDate = ''):
    dataday = DataAPI.MktEqudGet(tradeDate=u"",secID=u"",ticker=ticker,beginDate= begin,endDate= end,isOpen="1",field=u'',pandas="1")
    return dataday
```

```python
## 波动率计算函数
def ewm_vol(x,lam = 0.96):
    ewm = (np.square(x - x.mean())*np.array([lam**(len(x)+1-i) for i in range(1,len(x)+1)])).sum()
    return ewm
```

```python
df_1 = chs_factor()
```

```python
df_1['volity'] = df_1['chgPct'].rolling(30).apply(lambda x : ewm_vol(x,lam = 0.9))
```

```python
df_1['return'] = df_1['chgPct'].rolling(30).mean()
```

```python
plt.scatter(df_1['return'].values[200:300],df_1['volity'].values[200:300])
```

```python
def get_volity_return(ticker, lam = 0.96 ,n = 30):
    df_1 = chs_factor(ticker)
    df_1.index = list(df_1['tradeDate'])
    df_1['volity'] = df_1['chgPct'].rolling(n).apply(lambda x : ewm_vol(x,lam = lam))
    df_1['return'] = df_1['return'] = df_1['chgPct'].rolling(n).mean()
    return df_1[['ticker','return','volity']]
```
