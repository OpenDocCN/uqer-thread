# 分享：指定指数月跌涨统计

```python
from math import *
import numpy as np
import pandas as pd

pd.set_option('display.width', 200)
pd.set_option('display.precision', 2)
pd.set_option('colheader_justify', 'left')

def getMonthEnd(argBeginDate, argEndDate):
    data = DataAPI.TradeCalGet(exchangeCD=u"XSHG",beginDate=argBeginDate,endDate=argEndDate,
                field=["calendarDate", "isMonthEnd", "isWeekEnd"], pandas="1")
    
    data = data[data['isMonthEnd'] == 1]
    Dates = map(lambda x: x[0:4]+x[5:7]+x[8:10], data['calendarDate'].values.tolist())
    return Dates

def rptIndexByName(argIndexName, argBeginDate, argEndDate):
    indexData = DataAPI.MktIdxdGet(ticker=argIndexName, beginDate=argBeginDate, endDate=argEndDate,
        field=["secShortName","tradeDate","closeIndex"], pandas="1")
    indexData['yyyy'] = indexData['tradeDate'].apply(lambda x: x[0:4])
    indexData['mmmm'] = indexData['tradeDate'].apply(lambda x: x[5:7])
    
    idxRet = indexData.groupby(['yyyy','mmmm']).tail(1)
    idxRet['pctChg'] = idxRet['closeIndex'].pct_change()*100
    idxRet['pctChg'] = idxRet['pctChg'].apply(lambda x: round(x,2))
    idxRet = idxRet.dropna(axis=0,how='any')
    
    rptData = pd.DataFrame()
    for myKey, myGroup in idxRet.groupby(['yyyy']):
        myGroup = myGroup.set_index(['mmmm'])
        myRow = pd.DataFrame(myGroup['pctChg'].to_dict(),  index=[myKey])
        rptData = rptData.append(myRow)
    
    mySum = rptData.sum()
    myRow = pd.DataFrame(mySum.to_dict(), index=['sum'])
    rptData = rptData.append(myRow)
    
    return rptData

rptData = rptIndexByName('000300', '2006-12-01', '2017-01-01')
print rptData, "\n"
print rptData.describe()

#print getIndexByName('000016', '2016-01-01', '2017-11-01')
#print getIndexByName('000905', '2016-01-01', '2017-01-01')
```
