```python
import pandas as pd
import numpy as np

import matplotlib.dates as mdates
import matplotlib.pyplot as plt
```

```python
bond = ["530008", "270030", "519989", "070009"]
bond_name = [u'建信稳定增利C', u'广发聚财信用B', u'长信利丰', u'嘉实超短债']
stock = ['519983', '166011', '163412', '519156']
stock_name = [u'长信量化先锋', u'中欧盛世成长A', u'兴全轻资产', u'新华行业轮换配置A' ]
tickers_name = bond_name + stock_name
tickers = bond + stock
```

```python
history_nav = DataAPI.FundNavGet(dataDate=u"",secID=u"",ticker=tickers,beginDate=u"20150101",endDate=u"20160225",field=u"",pandas="1")
```

```python
nav_table = pd.pivot_table(history_nav, values='NAV', index=['endDate'], columns=['ticker'], aggfunc=np.sum)
```

```python
returns = nav_table.pct_change()
ts = returns.cumsum()
```

```python
weight = [0, 0, 0.3, 0.2, 0, 0.3, 0.2, 0]
portfolio_returns = returns.dot(weight)
portfolio = portfolio_returns.cumsum()
```

```python
fig = plt.figure(figsize = (25,8))
ax = fig.add_subplot(111)

t = mdates.datestr2num(ts.index.values.tolist())

ax.plot_date(t, ts.values, fmt='-')
ax.plot_date(t, portfolio, fmt='rH')

ax.legend(tickers + ['portfolio'], loc = 'best', fontsize = 14)
```
