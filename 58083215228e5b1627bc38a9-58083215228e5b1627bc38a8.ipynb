# 高管增持事件选股

高管增持事件是上市公司董事、监事和高级管理人员利用自有资金在二级市场增持自己公司的行为。因为高管作为“内部人”，肯定比外部人更深入的了解公司的实际运行情况，属于聪明的资金。高管增持的现象代表高管对公司的信心。

**关于高管增持的基本信息**：

> 高管大额增持通常是高管认为公司的股票受到低估

> 高管增持必须要在两个交易日内公告，基本都是先增持，后公告

> 高管增持通常发生在股价大跌以后



## 高管增持事件时间点确定
可以在东方财富的网站上看到每日高管增持的事件，这种公告一般都是交易时间之后再发布。比如[昆仑万维](http://data.eastmoney.com/executive/300418.html)在2016年10月11号定增，但是公告是在晚上8点多发出来的，所以最早也要在12号才能根据这个信息进行决策交易。分析的过程中要考虑到这个情况。

```python
# 该模块主要提供一些处理交易日的简单函数


def get_monthend_trading_day_list(begin_day="2007-01-01", end_day="2015-12-31"):
    """获取每月的最后一个交易日"""
    cal_dates = DataAPI.TradeCalGet(exchangeCD=u"XSHG", beginDate=begin_day, endDate=end_day, field="calendarDate,isMonthEnd")
    trading_days = cal_dates[cal_dates['isMonthEnd']==1]["calendarDate"].tolist()
    trading_days = [day[0:4] + day[5:7] + day[8:] for day in trading_days]  # 更改日期的格式，将'2014-01-03'转化为'20140103'
    return trading_days

def get_monthend(date_str='2016-01-04'):
    """获取该日的月底的最后一个交易日"""
    begin_date = datetime.strptime(date_str, '%Y-%m-%d')
    end_date = begin_date + 100 * timedelta(days=1)
    
    begin_day_str = begin_date.strftime('%Y-%m-%d')
    end_day_str = end_date.strftime('%Y-%m-%d')
    
    return get_monthend_trading_day_list(begin_day_str, end_day_str)[0]
    
def get_specific_trading_day(date_str='2016-01-04', window=20, direction='back'):
    """给定日期，可以通过这个函数获取前后N天的那个交易日"""
    date = datetime.strptime(date_str, '%Y-%m-%d')
    if direction=='back':
        begin_day = date - (window+5) * 4 * timedelta(days=1)   # 可以获取冗余的交易日，防止最终取不到有效数据
        end_day = date
    elif direction=='forward':
        begin_day = date
        end_day = date + (window+5)  * 4 * timedelta(days=1) 
        
    begin_day_str = begin_day.strftime('%Y-%m-%d')
    end_day_str = end_day.strftime('%Y-%m-%d')
        
    cal_dates = DataAPI.TradeCalGet(exchangeCD=u"XSHG", beginDate=begin_day_str, endDate=end_day_str, field="calendarDate,isOpen")
    trading_days = cal_dates[cal_dates['isOpen']==1]["calendarDate"].tolist()
    # trading_days = [day[0:4] + day[5:7] + day[8:] for day in trading_days]  # 更改日期的格式，将'2014-01-03'转化为'20140103'
    try:
        if direction=='back':
            return trading_days[-window]
        return trading_days[window]
    except:
        print date_str,window,direction


def get_trading_days_with_window(date_str='2016-01-04', window=20, direction='forward'):
    """给定日期，可以通过这个函数获取前后N天的交易日列表"""
    date = datetime.strptime(date_str, '%Y-%m-%d')
    if direction=='back':
        begin_day = date - (window+5)  * 4 * timedelta(days=1)
        end_day = date
    elif direction=='forward':
        begin_day = date
        end_day = date + (window+5)  * 4 * timedelta(days=1) 
        
    begin_day_str = begin_day.strftime('%Y-%m-%d')
    end_day_str = end_day.strftime('%Y-%m-%d')
        
    cal_dates = DataAPI.TradeCalGet(exchangeCD=u"XSHG", beginDate=begin_day_str, endDate=end_day_str, field="calendarDate,isOpen")
    trading_days = cal_dates[cal_dates['isOpen']==1]["calendarDate"].tolist()
    # trading_days = [day[0:4] + day[5:7] + day[8:] for day in trading_days]  # 更改日期的格式，将'2014-01-03'转化为'20140103'
    try:
        if direction=='back':
            return trading_days[-window:]
        return trading_days[:window]
    except:
        print date_str,window,direction
```

## 5. 高管增持事件在公告日T+1前后表现分析

1. 每个月末将所有股票按照市值大小分成20个组； 
2. 在每个交易日，以个股所在市值分组里的股票平均涨跌幅，作为其对应的市值基准； 
3. 计算出组合在各个交易日的实际累计净值； 
4. 计算出组合对应的市值基准组合在各个交易日的市值基准累计净值； 
5. 最后，组合的累计净值与其对应市值基准的累计净值比即为组合本身相对于其市值基准的超额收益，也即事件本身的真实alpha。


这个是paper中给的大体计算方法，按照个人的理解，每个交易日可能会有多只股票处于持仓，而且每天持仓可能会有变化。因为paper中每只股票的持仓周期为50个交易日。
在计算基准的时候，使用持仓对应分组所有的股票的日收益的平均值，并且每天加一后累乘，发现最终的结果和论文中的差距很大。基准和组合的收益几乎一样，并没有针对相应市值组合的超额收益。（可能是我的理解有误）下面是具体的代码和图形：

```python
# 对比基准的数据
bigger_300_df = raw_df[raw_df['change_money']>3000000]  # 增持金额大于300万的

bigger_300_df = bigger_300_df[bigger_300_df.index<'20160104']
bigger_300_df = bigger_300_df[['ticker','date']]
bigger_300_df.drop_duplicates(subset=['ticker','date'],inplace=True)   #去掉同一天重复增持的股票

my_dict = bigger_300_df.set_index('ticker')['date'].to_dict()
result_ls = []
for key in my_dict.keys():
    trading_days = get_trading_days_with_window(my_dict[key], 50, 'forward')[1:]  #向前取50天
    for day in trading_days:
        temp_ls = []
        temp_ls.append(key)
        temp_ls.append(day)
        result_ls.append(temp_ls)

df_2007_2016 = pd.DataFrame(result_ls,columns=['ticker','date'])
trading_day_ls = list(set(df_2007_2016['date']))
trading_day_ls.sort()

pro_ret_cum = 1.0  
base_ret_cum = 1.0
pro_ret_dict = {}  #保存组合收益
base_ret_dict = {}  # 保存市值基准收益

for day in trading_day_ls:
    temp_df = df_2007_2016[df_2007_2016['date']==day]
    all_tickers = list(set(temp_df['ticker']))
    portfolio_ret = DataAPI.EquRetudGet(listStatusCD=u"",secID=u"",ticker=all_tickers,beginDate=day,endDate=day,isChgPctl=u"",field=u"ticker,dailyReturnNoReinv",pandas="1")
    portfolio_ret = portfolio_ret.fillna(0.0)
    mkt = DataAPI.MktEqudGet(tradeDate=day,secID=u"",ticker=u"",beginDate=u"",endDate=u"",isOpen="",field=u"tradeDate,ticker,marketValue",pandas="1")
    mkt['lable'] = pd.qcut(mkt['marketValue'], 20, labels=False) 
    base_df = mkt[mkt['ticker'].isin(all_tickers)]
    all_group = list(set(base_df['lable']))

    base_tickers = list(set(mkt[mkt['lable'].isin(all_group)]['ticker']))
    base_ret = DataAPI.EquRetudGet(listStatusCD=u"",secID=u"",ticker=base_tickers,beginDate=day,endDate=day,isChgPctl=u"",field=u"ticker,dailyReturnNoReinv",pandas="1")
    base_ret = base_ret.fillna(0.0)
    pro_ret_cum = (1.0+np.mean(portfolio_ret['dailyReturnNoReinv'])) * pro_ret_cum
    pro_ret_dict[day] = pro_ret_cum
   
    base_ret_cum= (1.0+np.mean(base_ret['dailyReturnNoReinv']))* base_ret_cum
    base_ret_dict[day] = base_ret_cum

pro_ret_se = pd.Series(pro_ret_dict)
base_ret_se = pd.Series(base_ret_dict)

```

```python
fig = plt.figure(figsize=(12,5))
fig.set_tight_layout(True)
ax1 = fig.add_subplot(111)

    
pro_ret_index = pd.to_datetime(pro_ret_se.index)
base_ret_index = pd.to_datetime(base_ret_se.index)

ax1.plot(pro_ret_index, pro_ret_se.values,label=u'组合收益曲线')
ax1.plot(base_ret_index, base_ret_se.values,label=u'基准收益曲线')
ax1.legend(loc=2,prop=font)
ax1.set_title(u"高管增持策略组合相对市值基准", fontproperties=font, fontsize=16)
```

## 6. 高管增持事件在公告日前60天与后120天表现
 
> 下面代码主要使用增持金额大于300万的增持数据，可以重新设置初始条件，将增持金额的起始条件设为100万。
> 下面的代码大体思路(假设统计前60天表现)：

1. 分别读取发生增持的股票前60天到前4天的收盘价，计算这56天的收益 
2. 在60天前可以获得该股票对应的市值分组，获取这个分组的所有股票的60天的收盘价和4天前的收盘价，同样计算分组中每个股票的收益，然后取均值。
3. 个股减去基准即为超额收益，将每个股票的超额收益再取均值，即可得到下图：

```python
bigger_300_df = raw_df[raw_df['change_money']>3000000]  # 增持金额大于300万的
bigger_300_df = bigger_300_df[['ticker','date']]
bigger_300_df.drop_duplicates(subset=['ticker','date'],inplace=True)   #去掉同一天重复增持的股票

bigger_300_df['60days_ago'] = bigger_300_df['date'].apply(lambda x: get_specific_trading_day(x, 60,'back')) 
bigger_300_df['4days_ago'] = bigger_300_df['date'].apply(lambda x: get_specific_trading_day(x, 4,'back'))
bigger_300_df['t_plus_1'] = bigger_300_df['date'].apply(lambda x: get_specific_trading_day(x, 1,'forward'))
bigger_300_df['50days_later'] = bigger_300_df['date'].apply(lambda x: get_specific_trading_day(x, 50,'forward'))
bigger_300_df['120days_later'] = bigger_300_df['date'].apply(lambda x: get_specific_trading_day(x, 120,'forward'))



def get_ret(ticker, begin_day, end_day):
    """获取个股的一个时间段的收益"""
    begin_close = DataAPI.MktEqudAdjGet(tradeDate=begin_day,secID=u"",ticker=ticker,isOpen="",beginDate=u"",endDate=u"",field=u"closePrice",pandas="1")['closePrice']
    end_close = DataAPI.MktEqudAdjGet(tradeDate=end_day,secID=u"",ticker=ticker,isOpen="",beginDate=u"",endDate=u"",field=u"closePrice",pandas="1")['closePrice']
    try:
        ret = end_close[0] / begin_close[0] -1.0
        return ret
    except:
        print 'get_ret error: ' +ticker+begin_day+end_day
        return 999999999

def get_base_ret(ticker='000651', begin_day='2007-09-26', end_day='2007-12-19'):
    """获取一个股票对应的市值组合的一段时间的收益"""
    mkt = DataAPI.MktEqudGet(tradeDate=begin_day,secID=u"",ticker=u"",beginDate=u"",endDate=u"",isOpen="",field=u"tradeDate,ticker,marketValue",pandas="1")
    mkt['lable'] = pd.qcut(mkt['marketValue'], 20, labels=False) 
    base_df = mkt[mkt['ticker'].isin([ticker])]
    all_group = list(set(base_df['lable']))
    base_tickers = list(set(mkt[mkt['lable'].isin(all_group)]['ticker']))
    begin_close = DataAPI.MktEqudAdjGet(tradeDate=begin_day,secID=u"",ticker=base_tickers,isOpen="",beginDate=u"",endDate=u"",field=u"ticker,closePrice",pandas="1")
    
    end_close = DataAPI.MktEqudAdjGet(tradeDate=end_day,secID=u"",ticker=base_tickers,isOpen="",beginDate=u"",endDate=u"",field=u"ticker,closePrice",pandas="1")
    result_df = begin_close.merge(end_close, how='inner',on='ticker')
    result_df['ret'] = result_df['closePrice_y'] / result_df['closePrice_x'] -1.0
    return result_df['ret'].mean()
```

```python
def get_alpha(begin='60days_ago',end='4days_ago'):
    ret_dict = {}
    for year in range(2007, 2016):
        year_begin = str(year)+ '0101'
        year_end = str(year)+ '1231'
        year_df = bigger_300_df[(bigger_300_df.index<year_end)&(bigger_300_df.index>=year_begin)]

        alpha_ls = []
        for row in year_df.iterrows():
            ticker= row[1]['ticker']
            begin_day = row[1][begin]
            end_day = row[1][end]
            single_ret = get_ret(ticker, begin_day, end_day)
            if single_ret > 99999:
                continue
            bese_ret = get_base_ret(ticker, begin_day, end_day)
            alpha_ls.append(single_ret - bese_ret)
        ret_dict[str(year)]=np.mean(alpha_ls)

    return pd.Series(ret_dict)



back_60_se = get_alpha()
back_4_se = get_alpha('4days_ago','date')
forward_50_se = get_alpha('t_plus_1','50days_later')
forward_120_se = get_alpha('50days_later','120days_later')
```

```python
ind = back_60_se.index.astype(np.int64)
width = 0.2       # the width of the bars

fig = plt.figure(figsize=(16, 8))
ax = fig.add_subplot(111)

rects1 = ax.bar(ind-width, back_60_se.values, width, color='r')
rects2 = ax.bar(ind, back_4_se.values, width, color='royalblue')
rects3 = ax.bar(ind+width, forward_50_se.values, width, color='pink')
rects4 = ax.bar(ind+width*2, forward_120_se.values, width, color='lightskyblue')

ax.set_xticks(ind)
ax.set_xticklabels( [int(x) for x in ax.get_xticks()], fontproperties=font, fontsize=14)
# ax.tick_params(direction='out', pad=15)
ax.set_xlim(left=2006,right=2016)
# ax.spines['bottom'].set_position('zero')
# ax.xaxis.set_ticks_position('bottom')
ax.set_title(u"单日增持金额大于300万的前60个交易日与后120个交易日表现", fontproperties=font, fontsize=16)
ax.legend( (rects1[0], rects2[0], rects3[0],rects4[0]), ('[-60,4]', '[-4,0]', '[0,50]','[50,120]') )
```

## 7. 简单的高管增持的策略
通过上面的数据可以看到，在增持事件发生后立即买入，然后在50天以后卖出，会有较相于相同市值基准的超额收益。而且这种情况在2009年以后比较明显。但是120天以后卖出，收益的表现并不是特别稳定。


所以通过上面的结论写一个简单的策略：

1. 高管增持后的第一个可交易日买入等额的股票
2. 如果是开盘涨停，回测框架中不会成交，所以自动剔除了开盘涨停的股票
3. 持有50个交易日后卖出
4. 没有考虑到高管减持带来的影响
5. 参考基准为HS300，如果按照研报中ZZ500，效果比较差，囧～
