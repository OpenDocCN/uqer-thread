一财点睛今天节目嘉宾点评，选择9月最后一周中突破20日线并放量，并且前期（这里理解为100个交易日内）连续出现上穿20日均线并失败的（因为当日为突破均线行情，所以认为前几次都是失败的）

下面是用同花顺公式简单选取的：

> SC:= CROSS(C,MA(C,20)); //上穿20日均线

> CDC:= CROSS(C*1.02,MA(C,20)); //接近上穿20日均线

> LSSC:= SUM(CDC,100) ;//100日内历史上穿20日均线次数

> LBD:= V > MA(REF(V,1),5) * 1.5; //五日量比>2

> SELECT SC AND (LSSC > 2) AND LBD;
