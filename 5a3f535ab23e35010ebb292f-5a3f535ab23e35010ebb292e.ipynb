# 用python实现一个简单的因子分析框架

《量化投资策略:如何实现超额收益Alpha》一书中提到的分析框架非常简明，就是通过因子排序构造出的不同的投资组合，然后比较各个组合的表现是否和因子的排序一致。具体来说，对于单因子而言，通过根据将该因子排序后的股票五等分构成五个组合，然后看第一等分的组合的表现是否超过第二组合，依次类推；而对于多个因子而言，先根据第一个因子排序，对于第一等分的股票根据第二个因子排序并接着筛选出五分之一的股票，每个因子依次类推，同时对于其他分位的股票也是一样的过程。
这里以ROIC和现金流量比两个因子为例，以年报数据为基础，实现书中的分析框架。

由于财务报表存在追溯调整的情况，这里选择报表的Point In Time数据，这样使用的是当年最新的报表，避免使用后来追溯调整过的数据。

```python
import pandas as pd
import numpy as np
import dateutil
3
from datetime import datetime

tickerConverter = {'ticker':str}

stockMonthlyPriceData = pd.read_csv('stockMonthlyPriceData20171030.csv', converters = tickerConverter, encoding = 'GBK')

hs300IDXMonthlyPrice = pd.read_csv('hs300IDXMonthlyPrice20171031.csv', converters = tickerConverter, encoding = 'GBK')

stockInfoData = pd.read_csv('stocksInfo20171030.csv', converters = tickerConverter, encoding = 'GBK')

STFlagData = pd.read_csv('stockSTFlagData20171031.csv', converters = tickerConverter, encoding = 'GBK')
STFlagData = STFlagData[STFlagData.STflg != 'S']

secHaltData = pd.read_csv('secHaltData20171031.csv', converters = tickerConverter, encoding = 'GBK')
stockDelistedData = secHaltData[(secHaltData.assetClass == 'E') & (secHaltData.delistDate.notnull())]
```

从各个财务报表选取需要的财务数据用于计算一些财务因子。
这里以ROIC、现金收入比作为进行筛选的因子。

```python
"""
Data from the Income Statements
"""
stockISPointData = pd.read_csv('stockISPointData.csv', converters = tickerConverter, encoding = 'GBK')
stockAnnualISPointData = stockISPointData[stockISPointData.reportType == 'A']

stockAnnualISPointData.fillna(value = 0, inplace = True)
stockAnnualISPointData['NOPAT'] = stockAnnualISPointData.operateProfit - stockAnnualISPointData.incomeTax + (stockAnnualISPointData.NoperateIncome - stockAnnualISPointData.NoperateExp - stockAnnualISPointData.finanExp)*0.25

stockAnnualISFactors = stockAnnualISPointData[['endDate', 'ticker', 'publishDate', 'NOPAT', 'revenue']]

##删除部分数据，释放优矿的内存
del stockAnnualISPointData, stockISPointData
"""
Data from the Cashflow Statements
"""
stockCFPointData = pd.read_csv('stockCFPointData.csv', converters = tickerConverter, encoding = 'GBK')
stockAnnualCFPointData = stockCFPointData[stockCFPointData.reportType == 'A']

stockAnnualCFPointData.fillna(value = 0, inplace = True)
stockAnnualCFFactors = stockAnnualCFPointData[['endDate', 'ticker', 'publishDate', 'NCFOperateA']]
stockAnnualCFFactors.rename(columns={'publishDate' : 'CFPubDate'}, inplace = True)
stockAnnualCFFactors['hasCFData'] = 1 # The flag to determine whether the merged dataframe has data from CF statement

del stockCFPointData, stockAnnualCFPointData
```

在计算ROIC时，需要根据前一年的资产负债表计算IC，需要做增加一个去年的时间标签来拼接数据，当然也需要考虑报表追溯调整带来多个报表的问题。

```python
"""
Data from the past years financial reports
"""
#计算年初的因子值
#比如说当前会计报表日期是2016年12月31日，则最后披露时间应不晚于2017年4月底，
#则上年度会计报表日期为2015年12月31日，考虑到追溯调整的情况，最后更新时间也是2017年4月底
def latestDataLastYear(df):
    endDate = df.endDate.iloc[0]
    endDate = datetime.strptime(endDate, '%Y-%m-%d')
    
    endYear = endDate.year
    publishEndDate = datetime(endYear + 2, 5, 1).strftime('%Y-%m-%d')
    nextEndDate = datetime(endYear + 1, 12, 31).strftime('%Y-%m-%d')
    df['nextEndDate'] = nextEndDate
    df = df[df.publishDate < publishEndDate].sort_values('publishDate', ascending = False)
    return(df[~df.ticker.duplicated()])

stockBSPointData = pd.read_csv('stockBSPointData.csv', converters = tickerConverter, encoding = 'GBK')
stockAnnualBSPointData = stockBSPointData[stockBSPointData.reportType == 'A']            
lastYearBSReport = stockAnnualBSPointData.groupby('endDate', group_keys = False).apply(latestDataLastYear)
lastYearBSReport.fillna(value = 0, inplace = True)
lastYearBSReport['IC'] = lastYearBSReport['TShEquity'] + lastYearBSReport['LTBorr']

lastYearBSFactors = lastYearBSReport[['nextEndDate', 'ticker', 'IC']]

del stockBSPointData,stockAnnualBSPointData,lastYearBSReport
```

将来自不同报表的数据拼接到一块并计算一些财务因子。

```python
"""
Merge the data from different statements
"""
annualMergedData = stockAnnualISFactors.merge(lastYearBSFactors, left_on = ['endDate', 'ticker'], right_on=['nextEndDate', 'ticker'])
annualMergedData = annualMergedData.merge(stockAnnualCFFactors, how = 'left', left_on = ['endDate', 'ticker'], right_on =  ['endDate', 'ticker'])


annualMergedData['ROIC'] = annualMergedData.NOPAT / np.abs(annualMergedData.IC)
annualMergedData['CFINCOMERatio'] = annualMergedData.NCFOperateA / annualMergedData.revenue
```

由于Point In Time数据中包含了追溯调整的数据，所以需要对数据进行筛选，选出当年5月份之前发布的最新数据。

```python
#取每年5月1日之前的数据为年初的数据
def dataAtYearBeginning(df, sortDates):
    endDate = df.endDate.iloc[0]
    endDate = datetime.strptime(endDate, '%Y-%m-%d')
    endYear = endDate.year
    publishEndDate = datetime(endYear + 1, 5, 1).strftime('%Y-%m-%d')
    
    cond = (df.publishDate < publishEndDate)
    for sortDate in sortDates:
        cond = cond & (df[sortDate] < publishEndDate)
        
    df = df[cond].sort_values(sortDates, ascending = False)
    return(df[~df.ticker.duplicated()])

annualFactors = annualMergedData.groupby('endDate', group_keys = False).apply(dataAtYearBeginning, sortDates = ['publishDate', 'CFPubDate'])
```

由于按相关规定年报的最后发布日期是4月底，因此，这里根据每年4月份最后一个交易日对各个股票按因子进行排序并进行筛选，以当年5月1日到次年4月底为一个周期，并计算该周期内各个组合的回报率。

```python
#计算每年5月初至次年5月初的收益
def annualFactorReturn(df, monthlyPrice, factors, factorAscending):
    endDate = df.endDate.iloc[0]
    endDate = datetime.strptime(endDate, '%Y-%m-%d')
    endYear = endDate.year
    
    monthBegin = datetime(endYear + 1, 5, 1).strftime('%Y-%m-%d')
    endDateBegin = datetime(endYear + 1, 5, 31).strftime('%Y-%m-%d')
    endDateEnd = datetime(endYear + 2, 5, 1).strftime('%Y-%m-%d')
    
    lastDateEnd = max(monthlyPrice.endDate[monthlyPrice.endDate < monthBegin])
    endDateBegin = max(monthlyPrice.endDate[monthlyPrice.endDate <= endDateBegin])
    endDateEnd = max(monthlyPrice.endDate[monthlyPrice.endDate < endDateEnd])
    
    listedSec = stockInfoData[(stockInfoData.listDate < endDateBegin) & ((stockInfoData.delistDate.isnull()) | (stockInfoData.delistDate > endDateBegin))]
    
    beginPrice = monthlyPrice[monthlyPrice.endDate == endDateBegin][['ticker', 'tradeDays', 'openPrice']]
    beginPrice.rename(columns={'tradeDays' : 'tradeDays1'}, inplace = True)
    endPrice = monthlyPrice[monthlyPrice.endDate == endDateEnd][['ticker', 'closePrice', 'tradeDays']] 
    endPrice.rename(columns={'tradeDays' : 'tradeDays2'}, inplace = True)
    
    
    df = df.merge(beginPrice, how = 'left', on = 'ticker')
    #踢除还未上市的
    df = df[df.ticker.isin(listedSec.ticker)]
    
    #剔除停牌超过1个月的股票
    last1month = datetime.strptime(lastDateEnd, '%Y-%m-%d') + dateutil.relativedelta.relativedelta(months = -1)
    last1month = last1month.strftime('%Y-%m-%d')
    halt1montTicker = secHaltData[(secHaltData.haltBeginTime <= last1month) & ((secHaltData.haltEndTime >= lastDateEnd) | (secHaltData.haltEndTime.isnull()))].ticker
    df = df[~df.ticker.isin(halt1montTicker)]
    
    #踢除当日ST的股票
    STStocks = STFlagData[STFlagData.tradeDate == lastDateEnd].ticker
    df = df[~df.ticker.isin(STStocks)]
    
    df = df.merge(endPrice, how = 'left', on = 'ticker')
    df['return'] = (df.closePrice - df.openPrice)/df.openPrice
      
    #若当月一直停牌，无法建仓，对应资金的则不使用
    df.loc[df.tradeDays1 == 0, 'return'] = 0
    
    df.sort_values(factors, ascending = factorAscending, inplace = True)
      
    nfold = 5
    fold = np.arange(0, len(df) ,len(df)//nfold)
    fold[-1] = len(df)
    df['rank'] = range(len(df))
    cats = pd.cut(df['rank'], nfold, labels = False)
    df['cats'] = cats[:]
    df['sel'] = 1
    
    if type(factors) == list:
        if len(factors) > 1:
            for k in range(1, len(factors)):
                if type(factorAscending) == list:
                    ascending = factorAscending[k]
                else:
                    ascending = factorAscending
                for p in range(nfold):
                    fold = df[(df.cats == p) & (df.sel == k)].copy()
                    fold.sort_values(factors[k], ascending = ascending, inplace = True)
                    df.loc[fold.iloc[range(len(fold)//nfold)].index, 'sel'] = k + 1
        
    selID = len(factors)
    selDF = df[df.sel == selID]
    factorReturn = selDF[['cats', 'return']].groupby('cats').mean().T
    factorReturn['adjEndDate'] = endDateBegin
    factorReturn.set_index('adjEndDate', inplace = True)
    factorReturn['all'] = np.mean(df['return'])
    
    return(factorReturn)
    
annualReturns = annualFactors.groupby('endDate', group_keys = False).apply(annualFactorReturn, monthlyPrice = stockMonthlyPriceData, factors = ['ROIC', 'CFINCOMERatio'], factorAscending = False)
annualReturns.plot()


benchmarkRet = []
for k in range(len(annualReturns)):
    endDate = annualReturns.index[k]
    if k + 1 < len(annualReturns):
        nextYearEndDate = annualReturns.index[k + 1]
    else:
        nextYearEndDate = max(hs300IDXMonthlyPrice.endDate)
    openPrice = hs300IDXMonthlyPrice[hs300IDXMonthlyPrice.endDate == endDate]['openPrice'].iloc[0]
    closePrice = hs300IDXMonthlyPrice[hs300IDXMonthlyPrice.endDate == nextYearEndDate]['closePrice'].iloc[0]
    ret = closePrice/openPrice - 1
    benchmarkRet.append([endDate, ret])
    
benchmarkRet = pd.DataFrame(benchmarkRet)
benchmarkRet.columns = ['endDate', 'hs300']
benchmarkRet.set_index('endDate', inplace = True)

annualRtrnVSbenchmark = annualReturns.merge(benchmarkRet, left_index = True, right_index = True)
cumReturns = (annualRtrnVSbenchmark +  1).cumprod()
```
