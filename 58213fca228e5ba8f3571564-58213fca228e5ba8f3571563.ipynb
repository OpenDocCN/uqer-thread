下面介绍高维数据的匹配

```python
import numpy as np
import pandas as pd
def dist(aa,bb): #内部的距离的定义
    sum_temp=0
    q=len(aa)
    for i in range(q):
        sum_temp+=pow((aa[i]-bb[i]),2)#可以自己修改
    return pow(sum_temp,0.5)
def dtw(t,r):
    n=len(t)
    m=len(r)
    d=np.zeros((n,m))
    for i in range(n):
        for j in range(m):
            d[i][j]=dist(t[i],r[j])
    realmax=10000000
    D=np.ones((n,m))*realmax
    D[0][0]=d[0][0]
    #动态规划
    #print d
    #print D
    for i in range(1,n):
        for j in range(0,m):
            D1=D[i-1][j]
            if j>0:
                D2=D[i-1][j-1]
            else:
                D2=realmax
            if j>1:
                D3=D[i-1][j-2]
            else:
                D3=realmax
            D[i][j]=d[i][j]+min(D1,D2,D3)
            #print D
    return D[n-1][m-1]
def normalized(a):#为何要用标准化
    m=np.mean(a)
    s=np.std(a)
    return (a-m)/s
```

以下程序的架构是，给定now 和 day 以最后20天的数据作为待匹配的序列，在之前的历史中进行匹配。

```python
now="20161020"
day="20110101"
window_length=20
ticker_name='000001'
#name_type='preClosePrice'#股票前收盘价的名字
name_type=['closeIndex','turnoverValue']#指数收盘价的名字
#数据载入
#data=DataAPI.MktEqudGet(ticker=ticker_name,beginDate=day,endDate=now,field=[],pandas="1")#获得了历史上的股票数据
data=DataAPI.MktIdxdGet(ticker=ticker_name,beginDate=day,endDate=now,field=u"",pandas="1")#指数数据
#print data.head()
close=data[name_type].values #这里面的close实际上是二维数据（包含close和volume）
###########################################可以调用任何有用的数据研究
date=data['tradeDate'].values
distance=[]
begindate=[]
enddate=[]
x=close[-window_length:]#长度window_length的窗口 获得末尾数据
#数据标准化
t=normalized(x)
nowbegin=date[-window_length]#起始时间 指待匹配的序列的起始时间
nowend=date[-1]
for i in range(len(close)-window_length-4):#共有五个序列窗口在读数据
    begindate.append(date[i])
    enddate.append(date[i+window_length-1])#第一段 第二段 不断往后排
    r=normalized(close[i:i+window_length])#待匹配的样本历史每次循环都取出一段数据
    distance.append(dtw(t,r)) ###################距离认为是成交量和价格匹配的误差和 这个也可以改
print "done"
a=pd.DataFrame(begindate,columns=['begindate'])
b=pd.DataFrame(enddate,columns=['enddate'])
w=[distance]
w=np.transpose(w)#变成矩阵列向量
c=pd.DataFrame(w,columns=['distance'])#转成dataframe
a=a.join(b,how='outer')#ab先合成一起
record=a.join(c,how='outer')
record=record.sort(columns='distance')
print 'Begindate:  %s' %nowbegin
print 'Enddate:    %s' %nowend
print 'distance前五'
record.head()
#我们选前五画图
```

```python
top5=record.head()
price=np.zeros((6,window_length+5))
normprice=np.zeros((6,window_length+5))
initialA= np.empty((1,window_length+5,))
initialA[:] = np.NAN
norminitialA=np.empty((1,window_length+5,))
norminitialA[:]=np.NAN
close=DataAPI.MktIdxdGet(ticker=ticker_name,beginDate=day,endDate=now,field=u"",pandas="1")['closeIndex'].values
nclose=normalized(close[-window_length:])
for i in range(window_length):
    initialA[0][i]=close[-window_length+i]
    norminitialA[0][i]=nclose[-window_length+i]
for i in range(len(top5)):
    day=top5['begindate'].values[i]
#########也涉及到股票与指数的问题
    #ex=DataAPI.MktEqudGet(ticker=ticker_name,beginDate=day,endDate=now,field=[],pandas="1")#####股票形式
    ex=DataAPI.MktIdxdGet(ticker=ticker_name,beginDate=day,endDate=now,field=u"",pandas="1")####指数形式
#############################
    normtake=normalized(ex['closeIndex'].values[:window_length+5])
    take=ex['closeIndex'].values[:window_length+5]
    price[i]=take
    normprice[i]=normtake
price[-1]=initialA
normprice[-1]=norminitialA
price=np.transpose(price)
normprice=np.transpose(normprice)
price=pd.DataFrame(price,index=range(1,window_length+6),columns=['Top1','Top2','Top3','Top4','Top5','close'])
normprice=pd.DataFrame(normprice,index=range(1,window_length+6),columns=['normTop1','normTop2','normTop3','normTop4','normTop5','normclose'])
price
#正常价格
price.plot(figsize=(12,6),linewidth=1)
#标准化
normprice.plot(figsize=(12,6))
```
