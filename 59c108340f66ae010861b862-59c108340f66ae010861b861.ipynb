## **1. 中性化的一般做法**
我们一般做中性化，把需要中性掉的因子作为因变量，把行业哑变量和市值作为自变量，进行回归，取残差，用公式表示如下：
$$\vec{f_{t}}=\vec{I_{1,t}}+\vec{I_{2,t}}+...+\vec{I_{k,t}}+\vec{M_{t}}+\vec{\epsilon_{t}}$$
* $\vec{f_{t}}$表示t时刻因子在N个股票上的取值
* $vec{I_{k,t}}$表示t时刻第k个行业在N个股票上的取值，使用0，1哑变量进行表示
* $\vec{M_{t}}$表示t时刻市值取自然对数并进行横截面标准化之后（风险模型SIZE）的在N个股票上的取值
* 取$\vec{\epsilon_{t}}$作为中性化之后的因子值

有股票的数目比较多的行业，计算机行业，化工行业，机械设备，也有股票数目比较少的行业，比如说银行，钢铁。那么这样会引起很忙问题呢，我们知道，行业中性是使得不同行业之间的因子具有可比性，这是一个相对的概念，比如说，有的行业股票数量比较多，那么这种回归方法就会造成整个股票组合的多头组合其实会高估股票这个行业个股的Alpha值。来看个小例子：

```python
# EP = pd.read_csv('EP.csv', index_col=0)
EP
```

```python
date = EP.index[0]
date_ = ''.join(EP.index[0].split('-'))
pure_factor = pd.Series(standardize(neutralize(winsorize(EP.ix[date, :].to_dict()), 
                                                         target_date=date_, 
                                                         risk_module='short', industry_type='SW1', 
                                                         exclude_style_list=['BETA', 'RESVOL', 'MOMENTUM', 'EARNYILD', 'BTOP', 'GROWTH', 'LEVERAGE', 'LIQUIDTY'])))
```

```python
pure_factor.sort()
pure_factor.dropna(inplace=True)
universe1 = pure_factor.index[-100:].tolist()
```

```python
factor = pd.Series(winsorize(EP.ix[date, :].to_dict()))
factor.sort()
factor.dropna(inplace=True)
universe2 = factor.index[-100:].tolist()
```

```python
diff_univ = list(set(universe2)-set(universe1))
```

```python
data = DataAPI.EquIndustryGet(industryVersionCD=u"010303", secID=diff_univ, intoDate=date_, field=u"secID,industryName1",pandas="1")
fig = plt.figure(figsize=(16, 5))
ax = fig.add_subplot(111)
group= data.groupby('industryName1').count()
ax = group.plot(kind='bar', ax=ax)
labels = [label.decode("utf-8") for label in group.index.values] 
ax.set_xticklabels(labels, fontproperties=font, fontsize=13)
ax.legend(prop=font)
ax.set_xlabel(u'行业名称', fontproperties=font, fontsize=15)
ax.set_ylabel(u'股票数量', fontproperties=font, fontsize=15)
ax.set_ylim(0, 30)
ax.grid(True)
```

这说明了很多银行的股票，中性化完EP之后，我们基本不会去配。但是，今年这种情况是，EP越高就越好，不是相对的，而是绝对的，所以在今年这种行情不做行业中性反而会好一些。

```python
pre_date = '2017-01-26'
signal = neutralize(winsorize(EP.ix[pre_date, :].dropna().to_dict()), 
                                                         target_date=pre_date, 
                                                         risk_module='short', industry_type='SW1', 
                                                         exclude_style_list=['BETA', 'RESVOL', 'MOMENTUM', 'EARNYILD', 'BTOP', 'GROWTH', 'LEVERAGE', 'LIQUIDTY'])
    
    # 组合构建                
wts1 = pd.Series(long_only(signal, select_type=1, top_ratio=0.034, weight_type=1, target_date=''.join(pre_date.split('-')), universe_type='ZZ500')).to_frame()
signal = winsorize(EP.ix[pre_date, :].dropna().to_dict())
    
    # 组合构建                
wts2 = pd.Series(long_only(signal, select_type=1, top_ratio=0.034, weight_type=1, target_date=''.join(pre_date.split('-')), universe_type='ZZ500')).to_frame()
```

```python
data = DataAPI.EquIndustryGet(industryVersionCD=u"010303", secID=wts1.index.tolist(), intoDate=pre_date, field=u"secID,industryName1",pandas="1")
data = wts1.merge(data, how='left', left_index=True, right_on='secID')
fig = plt.figure(figsize=(16, 5))
ax = fig.add_subplot(111)
group= data.groupby('industryName1').sum()
ax = group.plot(kind='bar', ax=ax)
labels = [label.decode("utf-8") for label in group.index.values] 
ax.set_xticklabels(labels, fontproperties=font, fontsize=13)
ax.legend(prop=font)
ax.set_xlabel(u'行业名称', fontproperties=font, fontsize=15)
ax.set_ylabel(u'行业权重', fontproperties=font, fontsize=15)
ax.grid(True)
```

```python
data = DataAPI.EquIndustryGet(industryVersionCD=u"010303", secID=wts2.index.tolist(), intoDate=pre_date, field=u"secID,industryName1",pandas="1")
data = wts2.merge(data, how='left', left_index=True, right_on='secID')
fig = plt.figure(figsize=(16, 5))
ax = fig.add_subplot(111)
group= data.groupby('industryName1').sum()
ax = group.plot(kind='bar', ax=ax)
labels = [label.decode("utf-8") for label in group.index.values] 
ax.set_xticklabels(labels, fontproperties=font, fontsize=13)
ax.legend(prop=font)
ax.set_xlabel(u'行业名称', fontproperties=font, fontsize=15)
ax.set_ylabel(u'行业权重', fontproperties=font, fontsize=15)
ax.grid(True)
```

上面我们分开进行了回测，大体持仓数量控制在100只左右。从结果来看，是否进行行业中性化，对结果影响很大。看来还是要多多研究啊。
