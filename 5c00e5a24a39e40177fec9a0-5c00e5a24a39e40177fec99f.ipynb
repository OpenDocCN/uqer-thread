3. 我们以常见的技术指标，例如底背离（当股价指数逐波下行，而DIF及MACD不是同步下降，而是逐波上升，与股价走势形成底背离）来进行尝试

```python
# 底背离
def MACD_break(stock,begin='2017-01-01',end='2018-01-01'):

    fac = DataAPI.MktStockFactorsDateRangeGet(secID=stock,beginDate=begin,endDate=end,
                                              field=u"tradeDate,secID,MACD,DIFF,DEA",pandas="1")

    hist = DataAPI.MktEqudAdjGet(secID=stock,beginDate=begin,endDate=end,isOpen="1",
                                 field=u"tradeDate,secID,lowestPrice",pandas="1")

    data = pd.merge(fac, hist, on=['tradeDate','secID'], how='right')
     
    data['MACDup']=[1 if data.MACD[i] >= 0 else 0 for i in range(len(data))]

    method_break = lambda x: 1 if x[0] ==0 and x[1]==1 else 0
    if_break = pd.rolling_apply(data.groupby(by='secID').MACDup, 2, method_break)
    if_break.dropna(inplace=True)
    if_break = if_break.reset_index().set_index('level_1')

    data = pd.merge(data, if_break, left_index=True, right_index=True, how='left')
    data.drop(['secID_y','MACDup_x'],axis=1,inplace=True) 
    data.rename(columns={'secID_x':'secID','MACDup_y':'MACDup_break'}, inplace = True)
    data.set_index('tradeDate', inplace=True)

    # 选出MACD上穿的日期
    data_macd_bk = data[data['MACDup_break'] == True]

    choice_date_list = []
    date_list = list(data_macd_bk.index.values)

    for n in range(len(date_list)-2):
        now_d = date_list[n+1]
        next_d = date_list[n+2]
        last_now_d = cal.advanceDate(now_d,'-15B', BizDayConvention.Preceding).strftime('%Y-%m-%d')
        last_next_d = cal.advanceDate(next_d,'-15B', BizDayConvention.Preceding).strftime('%Y-%m-%d')

        # 取两个break前n天的数据
        data_case_now = data.loc[last_now_d:now_d]
        data_case_next = data.loc[last_next_d:next_d]

        # 取macd大于0的日期
        now_d_macd_list = []
        next_d_macd_list = []

        for date in data_case_now.index.values:
            if data_case_now.loc[date, 'MACD'] > 0:
                now_d_macd_list.append(date)

        for date in data_case_next.index.values:
            if data_case_next.loc[date, 'MACD'] > 0:
                next_d_macd_list.append(date)

        if len(now_d_macd_list) > 1:
            # now_d_macd_list[-1]为MACD的upbreak日期
            data_case_now = data_case_now[now_d_macd_list[-2]:]
        if len(next_d_macd_list) > 1:
            data_case_next = data_case_next[next_d_macd_list[-2]:]   

        # 取macd最低点的日期
        now_d_macd_min = 0
        next_d_macd_min = 0
        # 如果macd现在的低点 高于 过去的低点，取日期
        if data_case_now['DIFF'].min() <= data_case_next['DIFF'].min():
            now_d_macd_min = data_case_now['DIFF'].argmin()
            next_d_macd_min = data_case_next['DIFF'].argmin()

        if now_d_macd_min and next_d_macd_min != 0:
            # 如果价格峰值现在的低点 低于 过去的低点
            if data_case_now['lowestPrice'].min() > data_case_next['lowestPrice'].min():
                # 如果在next_d的macd—upbreak时刻，diff与dea都在0轴以下
                if data_case_next.loc[next_d,'DIFF'] < 0 and data_case_next.loc[next_d,'DEA'] < 0:
                    choice_date_list.append(next_d)

    if len(choice_date_list) > 0:            
        new_choice_date_list = [Date.parseISO(choice_date_list[0]),]

        for n in range(len(choice_date_list)-1):
            date1 = Date.parseISO(choice_date_list[n])
            date2 = Date.parseISO(choice_date_list[n+1])
            if date2 - date1 > 15L:
                new_choice_date_list.append(date2)
                
        # 信号出现后第二天开盘才能买入
        new_choice_date_list = [cal.advanceDate(x, '1B', BizDayConvention.Preceding) for x in new_choice_date_list]
        new_choice_date_list = [x.toISO() for x in new_choice_date_list]
        return new_choice_date_list
    else:
        return None
```

```python
start = '2018-01-01'                       
end = '2018-11-01'

cal_dates = DataAPI.TradeCalGet(exchangeCD=u"XSHG", beginDate=start, endDate=end).sort('calendarDate')
cal_dates = cal_dates[cal_dates['isOpen']==1]
trade_date_list = cal_dates['calendarDate'].values.tolist()
cal_dates = cal_dates[cal_dates['isMonthEnd']==1]
trade_month_list = cal_dates['calendarDate'].values.tolist()

universe = DynamicUniverse('ZZ500')
univ = []
for date in trade_month_list:
    univ = list(set(univ + universe.preview(date, skip_halted=True)))
```

```python
def get_quotation(start,end,universe):

    start_date = cal.advanceDate(start, '-30B', BizDayConvention.Preceding).strftime('%Y-%m-%d')
    end_date = cal.advanceDate(end, '30B', BizDayConvention.Preceding).strftime('%Y-%m-%d')
    df_eq = pd.DataFrame()
    
    for i in range(30,len(universe),30):
        df_tmp = DataAPI.MktEqudAdjGet(secID=universe[i-30:i],beginDate=start_date,endDate=end_date,
                                    field=u"secID,tradeDate,openPrice",isOpen=1,pandas="1")
        df_eq = pd.concat([df_eq,df_tmp],axis=0)
   
    df_tmp = DataAPI.MktEqudAdjGet(secID=universe[i:],beginDate=start_date,endDate=end_date,
                                field=u"secID,tradeDate,openPrice",isOpen=1,pandas="1")
    df_eq = pd.concat([df_eq,df_tmp],axis=0)
    return df_eq
```

```python
# 开盘到次日开盘的收益率
df_eq = get_quotation(start, end, univ)
df_eq_unstack = df_eq.set_index(['tradeDate','secID']).unstack()
open2open_yield = df_eq_unstack['openPrice'].pct_change().dropna(how='all')

chgpct = open2open_yield.copy()
chgpct = chgpct.stack()
chgpct = chgpct.reset_index().set_index('tradeDate')
chgpct.rename(columns={0:'chgPct'}, inplace=True)

# 平均收益率
chgpct_m = open2open_yield.copy()
chgpct_m['chgpct_mean'] = chgpct_m.mean(axis=1)
chgpct_m = chgpct_m[['chgpct_mean']]
```

```python
def signal_describe(date_decay='10'):
    # ============
    a=datetime.datetime.now()
    print ' data will be calculated for ' + str(len(univ)) + ' stocks:'
    count = 0
    secs_time = 0
    start_time = time.time()
    N = 30

    all_df = pd.DataFrame()
    for stk in univ: 
        # 信号的 endate 需要前推n天，因为下面需要后n天的数据
        last_d = cal.advanceDate(end,'-'+date_decay+'B', BizDayConvention.Preceding).strftime('%Y-%m-%d')
        date_list = MACD_break(stk,begin=start,end=last_d)

        if date_list is not None:
            case_df = pd.DataFrame(index=date_list, columns=['信号出现，后'+date_decay+'天的最大回撤','信号出现，后'+date_decay+'天的累计收益率','信号出现，基准的后'+date_decay+'天的累计收益率','信号出现，后'+date_decay+'天的累计超额收益率'])
            for date in date_list:
                # 后n天
                # stock
                next_d = cal.advanceDate(date,date_decay+'B', BizDayConvention.Preceding).strftime('%Y-%m-%d')
                case = chgpct[chgpct['secID']==stk]
                case = case.loc[date:next_d]
                case['cum_ret'] = case.chgPct + 1
                case['cum_ret'] = case.cum_ret.cumprod()
                df_cum_rets = case['cum_ret']
                running_min = np.minimum.accumulate(df_cum_rets)

                # beachmark
                b_case = chgpct_m.loc[date:next_d]
                b_case['cum_ret'] = b_case.chgpct_mean + 1
                b_case['cum_ret'] = b_case.cum_ret.cumprod()
                df_b_cum_rets = b_case['cum_ret']

                case_df.loc[date,'信号出现，后'+date_decay+'天的最大回撤'] = running_min[-1] -1
                case_df.loc[date,'信号出现，后'+date_decay+'天的累计收益率'] = df_cum_rets[-1] -1
                case_df.loc[date,'信号出现，基准的后'+date_decay+'天的累计收益率'] = df_b_cum_rets[-1] -1
                case_df.loc[date,'信号出现，后'+date_decay+'天的累计超额收益率'] = df_cum_rets[-1] - df_b_cum_rets[-1]

            case_df['secID'] = stk
            case_df.reset_index(inplace=True, drop=False)
            case_df.set_index(['secID','index'], inplace=True)
            all_df = pd.concat([all_df, case_df])

        count += 1
        if count > 0 and count % N == 0:
            finish_time = time.time()
            print count,
            print '  ' + str(np.round((finish_time-start_time) - secs_time, 0)) + ' seconds elapsed.'
            secs_time = (finish_time-start_time)

    b=datetime.datetime.now()
    print('finish！time cost: '+str(b-a))
    print ''
    return all_df
```

```python
all_df_10 = signal_describe(date_decay='10')
all_df_20 = signal_describe(date_decay='20')
```

```python
def signal_date(all_df, name):
    
    all_df.reset_index(inplace=True)
    signal = all_df[['secID','index']]
    signal.set_index('index', inplace=True)
    signal.sort_index(inplace=True)
    signal.reset_index(drop=False, inplace=True)
    
    signal.rename(columns={'index':'signal_date'}, inplace=True)
    signal['secID_signal_date'] = signal['secID'] +'_'+ signal['signal_date']
    signal['forecastType'] = name
    return signal
```

```python
def get_window_yield(unstack_yield,events,window):
    #获取事件窗口收益
    events = events.reset_index()
    
    def get_time_list(window,events):
    # 得到每个事件的时间窗口   
        period = str(window) + 'B'
        time_list = [[events['signal_date'][i],cal.advanceDate(events['signal_date'][i],period,BizDayConvention.Unadjusted)] for i in range(len(events['signal_date']))]
        return time_list
    
    time_list = get_time_list(window,events)
    offset = range(0,window+1,1)
    event_equity_list = []
    
    for i in range(len(events)):
        df_tmp = pd.DataFrame(unstack_yield.ix[time_list[i][0]:time_list[i][1].strftime('%Y-%m-%d'),events.iloc[i]['secID']].values,columns=[events.iloc[i]['secID_signal_date']],index = offset)
        event_equity_list.append(df_tmp)
        
    event_equity_window  = pd.concat(event_equity_list,axis=1)
    return event_equity_window
```

```python
def get_mean_outperform_yield(event_equity_window_yield,event_benchmark_window_yield):
    
    event_equity_window_cumyield = (1 + event_equity_window_yield).cumprod(axis=0) - 1
    event_benchmark_window_cumyield = (1 + event_benchmark_window_yield).cumprod(axis=0) - 1
    
    # 超额收益累计平均值
    event_outperform_window_yield = (event_equity_window_cumyield - event_benchmark_window_cumyield).mean(axis=1) 
    return event_outperform_window_yield
```

```python
def event_window_plot(open2open_yield,benchmarkYield,events,forecastType,window,label,legend,file):
    # 信号的时间窗口收益
    event_equity_window_yield = get_window_yield(open2open_yield,events[events['forecastType']==forecastType],window)
    event_benchmark_window_yield = get_window_yield(benchmarkYield,events[events['forecastType']==forecastType],window)
    event_outperform_window_yield = get_mean_outperform_yield(event_equity_window_yield,event_benchmark_window_yield)
    
    fig = plt.figure(figsize=(10,4))
    ax = fig.add_subplot(111)
    ax.plot(event_outperform_window_yield.index,event_outperform_window_yield.values,
            'dodgerblue',linewidth=2.5, label=label)
    plt.xlim(0,window+1)
    plt.ylim(event_outperform_window_yield.min(),event_outperform_window_yield.max())
    ax.set_title(legend, fontproperties=font, fontsize=18)
    ax.legend(fontsize=20)
    
    # 读取信号的统计文件
    des = pd.read_excel(file)
    des_ = pd.DataFrame(index=['信号数量','平均最大回撤','平均累计超额收益率','相对基准的胜率'],columns=['持有'+str(window)+'天'])
    
    des_.loc['信号数量','持有'+str(window)+'天'] = des.loc['count',u'信号出现，后'+str(window)+u'天的最大回撤']
    des_.loc['平均最大回撤','持有'+str(window)+'天'] = des.loc['mean',u'信号出现，后'+str(window)+u'天的最大回撤']
    des_.loc['平均累计超额收益率','持有'+str(window)+'天'] = des.loc['mean',u'信号出现，后'+str(window)+u'天的累计超额收益率']
    des_.loc['相对基准的胜率','持有'+str(window)+'天'] = des.loc['mean',u'相对基准的胜率']
    print '=================================='
    print '测试时间:',start,'-',end
    print '测试股票池: 中证500'
    print '=================================='
    print des_.to_html()
```
