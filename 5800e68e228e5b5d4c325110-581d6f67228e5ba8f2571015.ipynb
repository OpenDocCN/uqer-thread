- 移动平均线（MA）是技术分析中常用的一类趋势跟踪指标，其可以在一定程度上刻画股票价格或指数的变动方向。MA 的计算天数越多，平滑性越好，但时滞带来的延迟影响也越严重。因此，在使用 MA 指标进行趋势跟踪时，容易出现“跟不紧”甚至“跟不上”的情况。平滑性和延迟性在 MA 指标中成为了不可避免的矛盾，这就促使我们去寻找化解这一矛盾的工具和方法。 与 MA 类似的均线指标还有 EMA， 其本质是在计算中对靠近计算日的价格赋予更大的权重。EMA 指标的计算方式在信号处理理论中恰好对应着一类一阶低通滤波器，其可以将信号的高频分量进行有效的过滤。通过分析，我们认为如果希望滤波效果更好，则需要选择合适阶数的滤波器。上述一阶滤波的效果相对较差，通带和阻带间的过渡带太长；阶数越高，滤波器传输函数在截止频率附近衰减得越快，但同时通带会变得不平，也就是靠近截止频率的信号会有些放大。因此折中来看，可以选择二阶滤波器，我们根据二阶滤波器设计了 LLT 低延迟趋势线，发现其在低频部分的输出信号较强，同时相比 MA 均线和 EMA 均线，延迟大幅下降。

- LLT函数算法

```python
def LLT(closePrice,par):
    cp=copy.deepcopy(closePrice)
    tmp=copy.deepcopy(closePrice)
    a=2.0/(par+1.0)
    for i in np.arange(2,len(tmp)):
        tmp[i]=(a-(a*a)/4)*cp[i]+((a*a)/2)*cp[i-1]-(a-3*(a*a)/4)*cp[i-2]+2*(1-a)*tmp[i-1]-(1-a)*(1-a)*tmp[i-2]
    return tmp
    #return tmp[len(tmp)-1] # 看看series怎么取值
```

```python
#回测代码
# 可编辑部分与 strategy 模式一样，其余部分按本例代码编写即可

# -----------回测参数部分开始，可编辑------------

start = '2014-08-01'
end = '2016-08-01'
benchmark = 'HS300'                        # 策略参考标准
universe = set_universe('HS300','20140801')  # 证券池，支持股票和基金
capital_base = 1000000

# ---------------回测参数部分结束----------------

# 把回测参数封装到 SimulationParameters 中，供 quick_backtest 使用
sim_params = quartz.SimulationParameters(start, end, benchmark, universe, capital_base)
# 获取回测行情数据
idxmap, data = quartz.get_daily_data(sim_params)
# 运行结果
results = {}

# 调整参数快速回测
for x in range(len(Tuple)):
    

    refresh_rate = 1
    
    k1 = Tuple[x][0]
    k2 = Tuple[x][1]
    
    def initialize(account):
        account.sig={}  #记录持有还是卖出信号
        for i in universe:
            account.sig[i]=0
        pass
    def handle_data(account):
        price = account.get_attribute_history('closePrice',260)
        buy=[]
        sell=[]
        for stk in account.universe:
            
            MAk1 = LLT(price[stk][-260:],k1)
            MAk2 = LLT(price[stk][-260:],k2)
            
            #MAk1 = price[stk][-k1:].mean()
            #MAk2 = price[stk][-k2:].mean()
            
            if MAk1 >= MAk2:           
                if account.sig[stk] == 0:   #金叉买入
                    buy.append(stk)
                account.sig[stk]=1
            elif MAk1 < MAk2:
                if account.sig[stk] == 1:   #死叉卖出
                    sell.append(stk)
                account.sig[stk]=0
        if len(sell)>0:
            for j in sell:
                order_to(j,0)
        if len(buy)>0:
            for i in buy:
                order_pct_to(i, 1.0/len(buy))
       
        # 调用 account.my_parameter 进行计算或其他操作
        return

    # ---------------策略逻辑部分结束----------------

    # 把回测逻辑封装到 TradingStrategy 中，供 quick_backtest 使用
    strategy = quartz.TradingStrategy(initialize, handle_data)
    # 回测部分
    bt, acct = quartz.quick_backtest(sim_params, strategy, idxmap, data, refresh_rate = refresh_rate, max_history_window = 520)

    # 对于回测的结果，可以通过 perf_parse 函数计算风险指标
    perf = quartz.perf_parse(bt, acct)

    # 保存运行结果
    results[x] = perf
```

```python
#得到各个组合的风险指标
result = pd.DataFrame()
result['均线组合(快，慢)'] = Tuple
result['年化收益'] = [results[i]['annualized_return'] for i in range(len(Tuple))]
result['最大回撤'] = [results[i]['max_drawdown'] for i in range(len(Tuple))]
result['收益波动率'] = [results[i]['volatility'] for i in range(len(Tuple))]
result['信息比率'] = [results[i]['information_ratio'] for i in range(len(Tuple))]
pd.concat([result.head(10),result.tail(10)],keys=['前十行','后十行'],axis=0)
#画图
fig = plt.figure(figsize = (16,6))
ax = fig.add_subplot(111)
result.set_index('均线组合(快，慢)').plot(kind='bar',figsize=(22, 9),ax=ax,color=['r','y','#DB7093','b'],title=u'均线组合表现')
ax.legend(map(lambda x: x.decode('utf-8'),result.columns.values[1:]),prop=small_font,loc='best')
ax.set_xlabel(u'均线组合(快，慢)', fontproperties=small_font)
plt.xticks(fontproperties = small_font, rotation = 70)
ax.title.set_font_properties(font)
ax.set_axis_bgcolor('white')
ax.grid(color='gray', alpha=1.0, axis='y')
```
