```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
ss='20190214'
ee='20200703'

#negMarketValue流通市值
f=['secID','tradeDate','PB','vwap','negMarketValue']


# pool=['601788.XSHG','600030.XSHG','600837.XSHG','600999.XSHG','601211.XSHG']
pool=['000002.XSHE','600048.XSHG','601155.XSHG','600695.XSHG','000069.XSHE','001979.XSHE']#地产股
store_MarketValue=pd.DataFrame()
store_PB=pd.DataFrame()
store_vwap=pd.DataFrame()
'''构造dataframe，以tradedate为对齐标准(用其它个股时，需要保证无停牌，交易日长度一致)'''
for i in pool:
    tmp=DataAPI.MktEqudGet(secID=i,ticker=u"",tradeDate=u"",\
                           beginDate=ss,endDate=ee,isOpen="",field=f,pandas="1")
    tmp_MarketValue=tmp['negMarketValue']
    tmp_PB=tmp['PB']
    tmp_vwap=tmp['vwap']
    store_PB[i]=tmp_PB
    store_vwap[i]=tmp_vwap
    store_MarketValue[i]=tmp_MarketValue

store_MarketValue=store_MarketValue.div(store_MarketValue.sum(axis=1),axis=0)###
print(store_MarketValue.head())
print(store_PB.head())
#得到市值占比，然后以此为基础加权计算PB，vwap，
#即 marketvalue[0,0]*PB[0,0]+marketvalue[0,1]*PB[0,1]+... 

securities_PB=[]
securities_vwap=[]
for i in range (len(store_MarketValue)):
    tmp1=store_MarketValue.ix[i]
    tmp2=store_PB.ix[i]
    tmp3=store_vwap.ix[i]
    tt=0
    tt2=0
    for j in range (len(pool)):
        tt=tt+tmp1[j]*tmp2[j]
        tt2=tt2+tmp1[j]*tmp3[j]
    tt=round(tt,2)
    tt2=round(tt2,2)
    securities_PB.append(tt)

outcome=pd.DataFrame()
outcome['weighted_PB']=securities_PB
outcome['weighted_vwap']=securities_PB
print(outcome.head())

```
