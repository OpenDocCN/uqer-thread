1.2 技术指标因子使用的陷阱
---------------------


大部分技术指标的因子，现在的计算都是基于**未复权**的原始价格。而我们回测平台中使用的都是**复权**后的价格。所以，当使用这些技术因子的时候，要注意不要将因子和价格数据直接进行比较。

2. ``StockScreener``挑选股票池
----------------------------
***

为了帮助用户更好的获取满足条件的股票池，我们设计了``StockScreener``，使用方法：

    StockScreener(< 筛选条件表达式>)
    
返回值为满足条件的证券列表。

其中的删选条件表达式，可以有如下的形式：

    (Factor(’PE’).nlarge(100) | Factor(’PB’).pct_range(0.95, 1)) & Factor(’RSI’, keepvalue=True).value_range(70, 100)

> ``nlarge(n)``

筛选股票池中，值最大的``n``只股票。上式中表示选取``PE``最大的100支股票

> ``pct_range(a, b)``

筛选股票之中，按照百分比数排列，前``100*a%``到前``100*b%``的股票。上式中表示，筛选``PB``排名在95%到100%之间的股票。

> ``value_range(a, b)``

按照数值排列，大于a小于b的股票。上式中表示，筛选``RSI``指标大于70小于100的股票


2.1 一个完整的例子
-------------------

```python
from quartz.api import *

condition = (Factor('PE').nlarge(100) | Factor('PB').pct_range(0.95, 1)) & Factor('RSI', keepvalue=True).value_range(70, 100)
universe = StockScreener()
print(universe)
```
