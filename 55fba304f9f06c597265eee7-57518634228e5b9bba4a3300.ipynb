# 周一到周五的涨跌统计

最近有哥们问，周一涨跌概率是多少呢？立马打开了优矿，计算一下。

```python
import pandas as pd
data = DataAPI.MktIdxdGet(ticker='000001', field='tradeDate,closeIndex,preCloseIndex')
data['weekday'] = pd.DatetimeIndex(data['tradeDate']).weekday
data['indexChange'] = data['closeIndex'] > data['preCloseIndex']
data.dropna(inplace=True)
```

```python
week_days = data.groupby('weekday')['indexChange'].count()
#统计每个交易日的天数，0表示周一，1表示周二，类推。
week_days
```

```python
change_days = data.groupby('weekday')['indexChange'].sum()
# 上涨交易日的天数
change_days
```

```python
# 每个周几的上涨概率如下：
change_days/week_days
```
