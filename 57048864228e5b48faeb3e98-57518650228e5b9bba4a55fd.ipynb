# 海龟+羊驼 神兽拼接可行吗？

前几天在论坛讨论两个策略拼接在一起可行吗，我觉得想法很有意思，现就这个问题讨论整理出来，大家一起探讨探讨。
网友的想法是：
把海龟交易策略和羊驼策略结合起来，也就是说，买入策略按照海龟的交易法则买入10支，然后按照羊驼策略的每周选取表现差的剔除掉，再重新按照海龟买入法买入补充。
两个策略拼接的困难点在于买入卖出周期的匹配上，有可能：
1.卖出羊驼的当天没有达标的海龟买入
2.所持有的海龟寿数已尽，但我每隔一短时间才能卖出寿数已尽的海龟
以上两点，尤其是第二点可能会严重影响拼接策略的表现。
为印证这个想法，那我们就写个程序测试一下。为了避免上述第一个问题，我选择有买入信号时才卖出，保持10只的持仓数量。
注：以下代码参考了@陈亚龙 的《羊驼策略》和《侠之大者一起赚钱》。龙哥分享了很多有价值的干货，从他那里学到了很多，大家可以跑去围观。

策略表现并不好，分析原因可能出现在上面所述的第二条：所持有的海龟寿数已尽，但我每隔一短时间才能卖出寿数已尽的海龟。
于是就有了下面的这个策略，按照海龟策略买入，每天淘汰一只羊驼，不控制持仓数量，这样可以及时卖出寿数已尽的海龟。

结论：一个优质策略买入卖出必须是相匹配的，随意拼接可能会变成四不像。~~~海龟+羊驼，画面太美~~~
