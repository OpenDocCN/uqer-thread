# 按行业和因子选股的小工具

做一个小工具，非常简单，使用同花顺类似的行情交易软件都可以实现，之所以写呢，主要原因一是发个帖子增加社区活跃度，二是为了多敲敲Python代码。
工具的目的：**任意选择一个行业，再依据某一个或多个因子进行排序，输出选股列表**

# 1、提取股票基本行业信息
首先选择A股所有股票的行业分类信息，我选择了申万行业分类，其中申万行业一级分类共有28个，申万行业二级分类共有103个。当然，如果想选择其他行业分类标准，只需要修改industryVersionCD参数即可。

```python
# 设置股票池
universe = set_universe('A')

# 获取股票行业分类信息数据
df = DataAPI.EquIndustryGet(industryVersionCD=u"010303", secID=universe, field=u"secID,exchangeCD,secShortName,industry,isNew,industryID1,industryName1,industryID2,industryName2")
df = df[df.isNew == 1]  # 筛选出最新的行业分类
df.head(10)
```

申万行业一级分类共有28个，此处列出28个行业分类的名称：

```python
# 打印出一级行业分类名称及代码
industry1 = df.drop_duplicates("industryName1")
industry1 = industry1.loc[:,["industryName1","industryID1"]]
industry1
```

# 2、设置行业参数
工具的使用者需要在上面这些名称中**选择一个自己想要选择的行业名称作为参数sel_industry**，此处我以国防军工行业为例：

```python
# 选择某一行业
sel_industry = "国防军工"

df_sel_industry = df[df.industryName1 == sel_industry]  # 筛选出参数选择的行业
print df_sel_industry.sort("industryID2")
```

# 3、设置因子名称
**设置因子名称参数和日期参数**，输出选股列表。因子名称参数参考优矿提供的[因子名称](https://uqer.datayes.com/data/browse/101/101008/?page=1) 。日期参数是代表因子数据的日期。
此处以PE、PB两个因子为例：

```python
factor_name1 = "PE"
factor_name2 = "PB"
Date = "20170217"

# 行业选择出的股票代码
sel_universe = df_sel_industry.secID

# 获取因子某日的数据
sel_sec_factor_data = DataAPI.MktStockFactorsOneDayGet(tradeDate=Date,secID=sel_universe,field=u"secID,tradeDate,"+factor_name1+","+factor_name2)
# 将因子数据与股票信息链接
factor_sec_data = sel_sec_factor_data.merge(df_sel_industry,left_on="secID", right_on="secID", how="left")

# 将股票代码依据第一个因子排序，并输出股票列表
print factor_sec_data.sort([factor_name1])
```

# 4、自定义函数
将上面的过程，写成自定义函数的形式：

```python
def sel_industry_factor_sec(universe, sel_industry, factor_name1, factor_name2, Date="20170217"):
    '''
    功能：选择某一行业的股票，按因子从小到大排序输出。
    @universe：股票池
    @industry_name：一级行业分类名称
    @factor_name1:第一个因子的名称
    @factor_name2:第二个因子的名称
    @Date:因子数据日期，可空，默认是20170217，次日起参数必须是一个交易日期
    '''
    # 获取股票行业分类信息数据
    df = DataAPI.EquIndustryGet(industryVersionCD=u"010303",secID=universe,
                                field=u"secID,exchangeCD,secShortName,industry,isNew,industryID1,industryName1,industryID2,industryName2")
    df = df[df.isNew == 1]  # 筛选出最新的行业分类
    
    df_sel_industry = df[df.industryName1 == sel_industry]  # 筛选出参数选择的行业
    # 行业选择出的股票代码
    sel_universe = df_sel_industry.secID

    # 获取因子某日的数据
    sel_sec_factor_data = DataAPI.MktStockFactorsOneDayGet(tradeDate=Date,secID=sel_universe,field=u"secID,tradeDate,"+factor_name1+","+factor_name2)
    # 将因子数据与股票信息链接
    factor_sec_data = sel_sec_factor_data.merge(df_sel_industry,left_on="secID", right_on="secID", how="left")

    # 将股票代码依据第一个因子排序，并输出股票列表
    return factor_sec_data.sort([factor_name1])

# 调用自定义函数输出结果
universe = set_universe('A')
sel_industry = "汽车"
factor_name1 = "LFLO"  # 对数流通市值
factor_name2 = "PB"

data = sel_industry_factor_sec(universe, sel_industry, factor_name1, factor_name2)
print data
```
