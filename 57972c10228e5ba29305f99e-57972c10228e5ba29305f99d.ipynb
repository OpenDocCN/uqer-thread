关于问题,我的想法是这样的.
总的来说是做一个t检验.
先统计出了,每日的指数波动,百分比.
这些指数的波动百分比,就是随机不随机的判断.计算出均值和方差,这就是总体.
利用假设检验的思想,抽取了连续天数的样本来计算均值,计算出t值,用%5的显著水平.

从结果来看,貌似是在2天这种短线情况下,还是比较随机的,时间长了,就不太随机了.60天的情况下,貌似就不怎么随机了.
