### 因子那么多，怎样才是因子正确的打开姿势？
> 之前在 [因子预测能力全局观察](https://uqer.datayes.com/community/share/57c45649228e5b6d237c743f) 一文中尝试过一种方法，首先对所有股票的全部因子计算出 Rank-IC 值，然后再进行排序打分，最后找出更有预测能力的因子，但是由于因子种类众多，直接对比是否合理？带着这种疑虑思考，感觉应该还可以采取不同的思路来进行观察。后来看到 [因子那么多，怎么用才有效？](https://uqer.datayes.com/community/share/579ae6f5228e5ba28e05fe03) 一文后，决定采用文中的思路对因子进行更深入一点的分析。

### 因子的分组
> 在优矿的因子日报和周报中，将因子按照风格或经济学含义不同分为了不同的组合，比如：盈利因子、规模因子、成长因子、估值因子、现金流因子、营运力因子、分析师因子、收入因子、规模因子、技术因子等。因此决定沿用此分类方法对因子进行分组。

### 实践
> 吸收 [因子那么多，怎么用才有效？](https://uqer.datayes.com/community/share/579ae6f5228e5ba28e05fe03) 一文中的思路，首先通过对因子先进行分组，然后再根据相关性筛选出能够代表因子组合的代表因子。

> 由于原文中只举例分析了一只股票的情况，这里我们就需要尝试对A股所有股票进行整体测试来验证思路。

### 技术实现
> 实际测试中发现，虽然我申请了 3G 的内存，但是运行中还是出现了内存不足的现象。看来全A股共3000多只股票的因子数据获取并计算需要花费非常大的内存空间和时间，因此我采取了折衷方案设计，把所有股票按块分组，每次只计算块内股票的因子数据结果并保存，最后再把这些按块计算的结果合并成最终需要的数据。

```python
import numpy as np
import pandas as pd
import time
import seaborn as sns

DataAPI.settings.cache_enabled = True

def get_factor_dict():
    # 盈利因子组合
    revenue_factors = [ 'NetProfitRatio', 'OperatingProfitRatio', 'GrossIncomeRatio', 'ROA', 'ROA5', 'ROE', 'ROE5', 'EGRO', 'EPS', 'DilutedEPS' ]
    # 规模因子组合
    scale_factors = [ 'CurrentRatio', 'QuickRatio', 'DebtEquityRatio', 'FixAssetRatio', 'IntangibleAssetRatio' ]
    # 成长因子组合
    grow_factors = [ 'NetAssetGrowRate', 'TotalAssetGrowRate', 'OperatingRevenueGrowRate', 'OperatingProfitGrowRate', 'NetProfitGrowRate', 'NPParentCompanyGrowRate' ]
    # 估值因子组合
    evaluation_factors = [ 'PE', 'PS', 'PCF', 'PB', 'ASSI' ]
    # 现金流因子组合
    cash_factors = [ 'CashRateOfSales', 'CashToCurrentLiability', 'FinancingCashGrowRate', 'OperCashGrowRate', 'InvestCashGrowRate' ]
    # 营运力因子组合
    operation_factors = [ 'InventoryTRate', 'ARTRate', 'AccountsPayablesTRate', 'CurrentAssetsTRate', 'TotalAssetsTRate' ]
    # 分析师因子组合
    analyst_factors = [ 'REC', 'FY12P', 'SFY12P' ]
    # 技术因子组合
    tech_factors = [ 'MA5', 'MA10','MA20','MA60','MA120', 'VOL5',
                'VOL10', 'VOL20', 'VOL60', 'VOL120', 'EMA5', 'EMA10', 'EMA20', 'EMA60', 'EMA120',
                'REVS5', 'REVS10', 'REVS20','DAVOL5', 'DAVOL10', 'DAVOL20' ]
    styles = ['revenue', 'scale', 'grow', 'evaluation', 'cash', 'operation', 'analyst', 'tech']
    factor_dict = {'revenue': revenue_factors, 'scale': scale_factors, 'grow': grow_factors, 'evaluation': evaluation_factors,
        'cash': cash_factors, 'operation': operation_factors, 'analyst': analyst_factors, 'tech':tech_factors}    
    return factor_dict

class show_running_time(object):
  def __init__(self, count):
    self.count = count
    self.index = 0

  def __call__(self, func):
    def wrapped_f(*args, **kwargs):
      if self.index == 0:
        self.begin_time = time.time()
      result = func(*args, **kwargs)
      self.index += 1
      if self.index >= self.count:
        used_time = time.time() - self.begin_time
        print('%s : %f' % (func.__name__, used_time))
        self.index = 0
      return result
    return wrapped_f

#@show_running_time(1)
def MktStockFactorsDateRangeGet(secID, beginDate, endDate):
    return DataAPI.MktStockFactorsDateRangeGet(secID=secID, beginDate=beginDate, endDate=endDate)

@show_running_time(1)
def make_universe_corr_dfs(universe, factor_dict, beginDate, endDate):
    corr_dfs = {}
    for idx, secid in enumerate(universe):
        print('secid: %04d, %s' % (idx, secid))
        df = MktStockFactorsDateRangeGet(secID=secid, beginDate=beginDate, endDate=endDate) # 获取股票一段时间范围内的所有因子值
        corr_dict = {}
        for factor_group in factor_dict.keys():
            factors = factor_dict[factor_group] # 得到分组的因子列表
            factor_df = df[factors] # 取指定分组的因子值
            corr_df = factor_df.corr('pearson') # 计算分组的因子pearson相关系数
            corr_dict[factor_group] = corr_df # 保存
        corr_dfs[secid] = corr_dict # 保存当前股票的各分组的因子pearson相关系数
    return corr_dfs

# 读取索引值
# 如果运行时中途异常或者其他原因退出时，可以从中间结果重新开始
def read_idx():
    try:
        text = read('test1/slots_idx.txt') # 需要先在建立一个名为 test1 的目录
        return int(text)
    except:
        return 0

# 保存索引值  
def write_idx(idx):
    text = '%d' % idx
    write('test1/slots_idx.txt', text)
    
@show_running_time(1)
def make_universe_corr_dfs(universe, factor_dict, beginDate, endDate, block=100):
    count = len(universe)
    slots = (count + block - 1) / block # 分块计算结果
    start = read_idx() # 读取上一次运行时保存的索引值
    for idx in range(start, slots): # 开始分块计算结果
        begidx = idx * block
        endidx = (idx+1) * block
        if endidx > count:
            secids = universe[begidx:]
        else:
            secids = universe[begidx:endidx]
        slot_dict = {}
        for sidx, secid in enumerate(secids):
            print('secid: %04d, %04d, %04d, %s' % (idx, sidx, idx*block+sidx, secid))
            df = MktStockFactorsDateRangeGet(secID=secid, beginDate=beginDate, endDate=endDate) # 获取股票一段时间范围内的所有因子值
            corr_dict = {}
            for factor_group in factor_dict.keys(): # 按分组进行计算
                factors = factor_dict[factor_group] # 得到分组的因子列表
                factor_df = df[factors] # 取指定分组的因子值
                corr_df = factor_df.corr('pearson') # 计算分组的因子pearson相关系数
                corr_dict[factor_group] = corr_df # 保存pearson相关系数
            slot_dict[secid] = corr_dict # 保存当前股票的各分组的因子pearson相关系数
        slot_dfs = pd.DataFrame(slot_dict) # 转换成 DataFrame
        slot_dfs.to_hdf('test1/slots_%04d.hdf' % idx, 'df') # 保存到 test1 目录下
        write_idx(idx+1) # 保存索引值
        
def test():
    beginDate='2012-01-01'
    endDate='2016-08-29'
    universe = set_universe('A', beginDate)
    #universe = universe[-50:-20]
    factor_dict = get_factor_dict()
    make_universe_corr_dfs(universe, factor_dict, beginDate, endDate, 100)

test()
print('game over!')
```

### 合并中间结果
> 有了中间结果，现在可以合并成最终的数据了。

```python
def combine_universe_corr_dfs(universe, block=100):
    count = len(universe)
    slots = (count + block - 1) / block
    corr_dfs = pd.DataFrame()
    for idx in range(slots):
        slot_dfs = pd.read_hdf('test/slots_%04d.hdf' % idx, 'df')
        corr_dfs = pd.concat([corr_dfs, slot_dfs], axis=1)
    return corr_dfs

def test():
    beginDate='2012-01-01'
    endDate='2016-08-29'
    universe = set_universe('A', beginDate)
    return combine_universe_corr_dfs(universe, 100)
    
corr_dfs = test()
print(corr_dfs.shape)
print('game over!')    
```

### 如何进行整体观测？
> 现在我们已经有了每只股票按照因子进行分组的pearson相关系数矩阵，下一步就是我们如何来整体观测整个A股所有股票的pearson相关系数矩阵？

> 最合适的方法我并不了解，不过考虑到相关系数矩阵的每个单元都代表一个因子与另一个因子的相关系数值，因此我们可以按照矩阵的每个单元对所有股票进行相加求和后求均值，就可以得到一个值，代表所有股票两两因子之间的相关系数均值。

```python
def make_universe_avg_dfs(corr_dfs, universe, factor_dict):
    corr_dfx = {}
    avg_dfs = {}
    for factor_group in factor_dict.keys():
        factor_list = factor_dict[factor_group] # 得到分组因子列表
        factor_cnt = len(factor_list) # 得到当前分组的因子总个数
        cnt_df = pd.DataFrame(np.zeros((factor_cnt, factor_cnt)), index=factor_list, columns=factor_list) # 用来对DataFrame中非缺失值出现的次数进行计数
        sum_df = pd.DataFrame(np.zeros((factor_cnt, factor_cnt)), index=factor_list, columns=factor_list) # 用来对DataFrame中每个单元进行求和
        for secid in universe:
            x = corr_dfs[secid].loc[factor_group] # 取出当前股票指定分组的因子pearson相关系数
            if x.empty: # 如果DataFrame为空，则表示所有的值缺失
                continue
            assert(x.shape == (factor_cnt, factor_cnt)) # 断言取出的DataFrame矩阵大小匹配
            sum_df += x.fillna(0) # 首先把NaN值替换成0，然后累计求和（这是由于DataFrame进行加法运算时，如果出现NaN值，则相加的结果也为NaN）
            tmp_x = x.copy() 
            tmp_x[~pd.isnull(tmp_x)] = 1 # 然后复制一份x出来，替换所有的非缺失值为1
            cnt_df += tmp_x.fillna(0) # 对非缺失值出现的次数进行计数
        #print(sum_df.to_html())
        cnt_df[abs(cnt_df) < 0.001] = np.nan # 为了防止出现除零错误，因此对于计数值为零的单元设置为NaN
        print(cnt_df.to_html())
        avg_df = sum_df / cnt_df
        print(avg_df.to_html())
        avg_dfs[factor_group] = avg_df
    return avg_dfs

def test(corr_dfs):
    beginDate='2012-01-01'
    endDate='2016-08-29'
    universe = set_universe('A', beginDate)
    #universe = universe[-30:-20]
    factor_dict = get_factor_dict()    
    return make_universe_avg_dfs(corr_dfs, universe, factor_dict)

avg_dfs = test(corr_dfs)
print('game over!')
```

### 因子分组后的筛选
> [因子那么多，怎么用才有效？](https://uqer.datayes.com/community/share/579ae6f5228e5ba28e05fe03) 一文中使用 factor_filter 函数对因子进行筛选，当多个因子相关性非常高时，会自动选择因子列表中排序在最前面的因子作为分组中的代表因子，而自动筛选掉其他因子。

> 为了更方便观察哪些因子因为相关性非常高而被筛选掉，我们重新设计了一个新的函数 factor_filter_groups 来进行观察。

```python
def factor_filter_groups(corr_df, threshold = 0.6):
  corr_arr = corr_df.as_matrix()
  arr = np.tril(corr_arr - np.diag(corr_arr.diagonal()))
  mask = abs(arr) > threshold
  f_filter = []
  groups = []
  for i in range(mask.shape[0]):
    if i in f_filter:
      continue
    col = mask[:, i]
    if np.any(col):
      group = np.where(col)[0].tolist()
      f_filter.extend(group)
      group.insert(0, i)
      groups.append(group)
    else:
      groups.append([i])
  result = []
  for group in groups:
    result.append([corr_df.index[s] for s in group])
  return result

def test(avg_dfs):
  factor_dict = get_factor_dict()
  for factor_group in factor_dict.keys():
    corr_df = avg_dfs[factor_group]
    result = factor_filter_groups(corr_df)
    print('%s: %s' % (factor_group, result))
    
test(avg_dfs)
```

### 简单的说明
> 可以看到以规模因子组合为例，其中 CurrentRatio 、QuickRatio、DebtEquityRatio 因子相关性非常高，其中 CurrentRatio 因子还会作为代表因子被选出，而 QuickRatio、DebtEquityRatio 因子则会被筛选掉。而 FixAssetRatio 和 IntangibleAssetRatio 因子与前面的因子相关性都较低，因而都保留。最终被选出的因子为 CurrentRatio、FixAssetRatio、IntangibleAssetRatio 因子。我们也可以运行原 factor_filter 函数来验证一下。

```python
def factor_filter(corr_df, threshold = 0.6):
    f_filter = []
    for col,row in [(s, t) for s in range(len(corr_df)) for t in range(s+1,len(corr_df))]:
        if (col in f_filter) or (row in f_filter):
            continue
        if abs(corr_df[corr_df.index[col]][corr_df.index[row]]) > threshold:
            f_filter += [row]
    return [corr_df.index[s] for s in range(len(corr_df)) if not s in f_filter]

def test(avg_dfs):
  factor_dict = get_factor_dict()
  for factor_group in factor_dict.keys():
    corr_df = avg_dfs[factor_group]
    result = factor_filter(corr_df)
    print('%s: %s' % (factor_group, result))
    
test(avg_dfs)
```

### 一点总结
> 这样首先把因子根据经济学意义以及风格进行分组之后，再根据因子相关性的高低可以进一步划分出了不同的分组，然后同组的因子无论从经济学意义还是相关性上考虑都更加具有相似性，再进行预测能力或者其他的对比从而筛选出最有代表性的因子，也似乎更合理一些。
