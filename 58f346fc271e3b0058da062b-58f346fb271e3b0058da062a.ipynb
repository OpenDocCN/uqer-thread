# 28风格转换量化策略实践

今天无意中发现一篇文章《分享一套量化模型》 http://www.360doc.com/content/15/1108/18/28745347_511701259.shtml  ，原作者不详，策略也是简单粗暴，就是统计这两个指数过去20天的涨跌幅平均值，谁涨的多就开谁的多单，谁涨的少就开谁的空单。

　　1、当过去20天均涨幅，上证50>中证500且>0,全仓买入510050

　　2、当过去20天均涨幅，中证500>上证50且>0,全仓买入510500

　　3、当过去20天均涨幅，中证500和上证50都<0，空仓。

按原文所述在A股能达到23%的年化率。最近正在优矿学习写量化程序，就拿这个策略实践了下。

拿510500和510050分别替代原文中的上证50和上证500，从2013年开始测试，结果与原文年化率基本一致。

但是也发现些问题，自2015年下半年开始这个策略就一直在回撤，所以策略应该还需要改进优化。
