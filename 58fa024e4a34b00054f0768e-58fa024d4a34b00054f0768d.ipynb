```python
###g3minr跌幅最大,g4liuchang下跌走势流畅
####1.初始设置：导入模块、设置日期、股票池等
import pandas as pd
import datetime as dt
universe2=set_universe('A')
# universe=set_universe('HS300')#先以沪深300为股票池
xingu=DataAPI.EquGet(secID=universe2,ticker=u"",listStatusCD=u"",field=['secID','secShortName','listDate'],pandas="1")
xingu=xingu.ix[xingu.listDate>'2016-09-23',:]
for stk in xingu.secID:
    if stk in universe2:
        universe2.remove(stk)
# 抓取最低价的日期区间
end='20161205'
begin='20161128'
list=[]
highstlist=pd.DataFrame()
```

```python
####2.在设定时间段内，获取股票池所有股票的日行情信息,存在price2里  
# for stock in universe2:
#     price2=DataAPI.MktEqudAdjGet(secID=stock,beginDate=begin,\
#                                  endDate=end,field=['ticker','secShortName',\
#                                                     'tradeDate','highestPrice','closePrice'],pandas='1')
#     # price.index=price['ticker']
#     # price.drop('ticker',axis=1,inplace=True)

# ## 储存最高价及其最高价出现的交易日期,存在highstlist
#     if len(highstlist)==0:
#         highstlist=price2[price2.highestPrice==price2.highestPrice.max()]
#     else:
#         highstlist=highstlist.append(price2[price2.highestPrice==\
#                                             price2.highestPrice.max()],ignore_index=True)
# highstlist.drop_duplicates(subset='ticker',keep='last',inplace=True)

price2=DataAPI.MktEqudAdjGet(secID=universe2,beginDate=begin,\
                            endDate=end,field=['ticker','secShortName','tradeDate',\
                                               'highestPrice','closePrice'],pandas='1')
price2.sort(columns=['ticker','highestPrice'],ascending=[False,False],inplace=True)
highstlist=price2.drop_duplicates(['ticker'],keep='first')
highstlist.index=range(0,len(highstlist))
# highstlist=highstlist.ix[highstlist.ticker!='000029',:]
        
```

```python
# ####3.借鉴最大回撤指标，计算上述股票从最低价至今的最大回撤，并存入infor2。

####3.借鉴最大回撤指标，计算上述股票在给定区间的最大回撤，并存入infor。
infor2=highstlist.ix[:,0:4]
infor2['drawback']=0
today=dt.datetime.today().strftime('%Y%m%d')##先转为Timestamp，再转为yyyymmdd字符型日期
for i in infor2.index:
    # date=pd.to_datetime(lowestlist.ix[i,'tradeDate']).strftime('%Y%m%d')
    # data=DataAPI.MktEqudAdjGet(ticker=lowestlist.ix[i,'ticker'],beginDate=date,endDate=today,\
    #                            pandas='1',field=['ticker','secShortName','tradeDate','closePrice'])
    data=DataAPI.MktEqudAdjGet(ticker=highstlist.ix[i,'ticker'],\
                               beginDate=pd.to_datetime(highstlist.ix[i,'tradeDate'])\
                               .strftime('%Y%m%d'),endDate=end,pandas='1',\
                               field=['ticker','secShortName','tradeDate','highestPrice',\
                                      'lowestPrice','closePrice'])
    data['high']=0
    data['huiche']=0

    for j in range(0,len(data.tradeDate),1):
        # lowdate=lowestlist.ix[j,'tradeDate']
        # data.ix[j,'high']=data.ix[data.tradeDate<data.ix[j,'tradeDate'],'highestPrice'].max()
        low=data.ix[data.tradeDate<data.ix[j,'tradeDate'],'lowestPrice'].min()
        # print (high)
        # print (data.ix[i,3])
        data.ix[j,'huiche']=1-(data.ix[j,'highestPrice']/low)
    infor2.ix[i,'drawback']=data.ix[:,'huiche'].max()
        
    
```

```python
####4.处理区间最大跌幅
infor2['return']=0
infor2['minre']=0
for i in infor2.index:
    data2=DataAPI.MktEqudAdjGet(ticker=highstlist.ix[i,'ticker'],\
                               beginDate=pd.to_datetime(highstlist.ix[i,'tradeDate']).strftime('%Y%m%d')\
                               ,endDate=end,\
                               pandas='1',field=['ticker',\
                                                 'tradeDate','highestPrice'\
                                                 ,'lowestPrice','closePrice'])
    a=len(data2)-1
    minp=data2.closePrice.min()
    maxp=infor2.ix[i,'highestPrice']
    minre=(minp/maxp)-1
    re=(data2.ix[a,'closePrice']/maxp)-1
    infor2.ix[i,'minre']=minre
    infor2.ix[i,'return']=re
    
```

```python
###5.后续处理
infor2['r/d']=infor2['return']/infor2.drawback
infor2['mr/d']=infor2.minre/infor2.drawback
result11=infor2.ix[infor2.highestPrice!=0,:]
result12=result11.ix[:,:]
# result2=result1.ix[result1.maxre<=1,:]
result12=result12.ix[result12.drawback>0,:]
# result4=result3.sort(columns=['drawback','maxre','return'],axis=0,ascending=[True,False,False])
g3minr=result12.sort(columns=['return'],axis=0,ascending=[True])
g4liuchang=result12.sort(columns=['r/d'],axis=0,ascending=[True])
# result5=result4.ix[result4.tradeDate<='2017-01-31',:]
# result5
```
