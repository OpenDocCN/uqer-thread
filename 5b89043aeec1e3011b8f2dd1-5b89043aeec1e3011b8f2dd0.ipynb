数据处理

```python
import pandas as pd
import numpy as np
    
data = pd.read_csv('QFII.csv')          #修改文件名字
data['CGChange'][data['CGChange']=='增持'] = 1
data['CGChange'][data['CGChange']=='新进'] = 2
data['CGChange'][data['CGChange']=='减持'] = -1
data['CGChange'][data['CGChange']=='持平'] = 0
data 

for idx in range(len(data['date'])):
    tmp_id = str(data.ix[idx,'SCode']).zfill(6)
    if tmp_id[0] == '6':
        tmp_id = tmp_id + '.XSHG'
    if tmp_id[0] == '0' or tmp_id[0] == '3':
        tmp_id = tmp_id + '.XSHE'
    data.ix[idx,'SCode'] = tmp_id
    data.ix[idx,'date'] = data.ix[idx,'date'][:7]
data.to_csv('all_data.csv')
data.head()
```
