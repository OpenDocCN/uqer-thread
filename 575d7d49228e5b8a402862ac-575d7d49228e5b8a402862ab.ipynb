未来股价涨跌对于当前的我们来说，是一种未知，即“不确定性”，从信息论的角度，我们弄清楚一件不确定性高的事情就需要更多的信息，弄清一件不确定性低的事情只需要少量的信息。香农（Claude Shannon）用信息熵来衡量不确定性的大小，其公式为$H(x) = \sum p(x) log\frac{1}{p(x)}$

举例来看，抛一枚硬币，信息熵H(x) = $\frac{1}{2} log 2 +\frac{1}{2} log 2 = 1$ ,即，我们只需要1bit的信息就可以知道正面还是反面向上，比如有人告诉我True(二进制的1)

未来股价涨跌也是两种可能，我们可以计算出这种不确定性的信息熵。

同样，我们也需要很多的信息来消除不确定性，优矿上提供了丰富的历史数据，新闻数据，社交数据，这些信息量很大，但是对股价预测这件事的用处大不大？哪一种指标对预测的效果更好？

香农提出了一个概念“互信息（Mutual information）”，$I(X;Y) = \sum_{x\in X,y\in Y}P(x,y)log \frac{p(x,y)}{p(x)p(y)}$。 如果互信息大，说明已知的信息对不确定性的消除帮助很大，反之帮助很小。

下面，我想从一些历史数据中，找出对股价涨跌的预测帮助更大的信息，即互信息更大。

x:当月股价涨幅
y:上月换手率周平均值

### **两者“互信息”**
$$I(x;y) = \sum p(x,y)log{\frac{p(x,y)}{p(x)p(y)}}$$

```python
security = universe[0]
data=DataAPI.MktEqumGet(secID=security,ticker=u"",monthEndDate=u"",isOpen=u"",beginDate=beginDate,endDate=endDate,field=u"",pandas="1")
x = np.array(data['chgPct'])
y = np.array(data['avgTurnoverRate'])
x = x[::-1][1:]
y = y[::-1][:-1]
```

对$p(x) ,p(y) , p(x,y)$进行统计：将x,y分别划分为2档，分别赋值为（0,1），即知道y所在区间，对于预测x所在区间是否有帮助

```python
#我们把x,y改为离散的数值，方便统计
x = x>0.05
x = x+0

y = y>y.mean()
y=y+0

A = np.column_stack([x,y])
trans = np.array([1,2])
A =A.dot(trans)
# 统计频率
from collections import Counter
count_x = Counter(x)
count_y = Counter(y)
count_xy = Counter(A)
```

分别计算$p(x),p(y),p(x,y)$

```python
num = len(x)
px ={}
px[0] = round(float(count_x[0])/num,2)
px[1] = round(float(count_x[1])/num,2)


py={}
py[0] = round(float(count_y[0])/num,2)
py[1] = round(float(count_y[1])/num,2)

pxy={}
pxy[0] = round(float(count_xy[0])/num,2)
pxy[1] = round(float(count_xy[1])/num,2)
pxy[2] = round(float(count_xy[2])/num,2)
pxy[3] = round(float(count_xy[3])/num,2)
```

计算信息熵：$H(X) = \sum p(x) log \frac{1}{p(x)}$

计算互信息：$I(x;y) = \sum p(x,y) log{\frac{p(x,y)}{p(x)p(y)}}$

```python
hx = px[0]*log(1/px[0],2)+px[1]*log(1/px[1],2)

ixy = 0
for i in [0,1]:
    for j in [0,1]:
        ixy += pxy[i*1+j*2]*log(pxy[i*1+j*2]/(px[i]*py[j]))
print "信息熵H(X) = %s，互信息I(X;Y) = %s" %(hx,ixy)
```

我们可以发现互信息很小，对股价涨跌的预测帮助并不大。
