```python
#原始因子数据
universe = set_universe('SH50') 
factor_list=['CMRA','CTOP','DAVOL5','EGRO','EMA10','EPS','HBETA','LCAP','MA10','REVS10','PB','PE']
Style_factor = DataAPI.MktStockFactorsOneDayGet(tradeDate="20200323",secID=universe,field=['secID']+factor_list,pandas="1")
Style_factor.fillna(0,inplace = True)
Style_factor
```

```python
#normalize化
market_value = np.exp(Style_factor.LCAP)
weight = market_value/np.sum(market_value)
normalized_style = Style_factor[factor_list].apply(lambda x:(x-x.dot(weight))/np.std(x))
normalized_style
```

```python
#行业因子
industry_list = ['secID','secShortName','industryName1']
Industry_factor = DataAPI.EquIndustryGet(secID=universe,industryVersionCD="010303",
                                intoDate="20200323",field=industry_list,pandas="1")
Industry = pd.merge(Industry_factor,pd.get_dummies(Industry_factor.industryName1),left_index = True,right_index = True)
del Industry['industryName1']
Industry['cons'] = 1
Industry.columns = ['secID','secShortName','交通运输','休闲服务','化工','医药生物','国防军工','家用电器','建筑材料',
 '建筑装饰','房地产','有色金属','机械设备','汽车','电子','电气设备','通信',
 '采掘','银行','非银金融','食品饮料','cons']
Industry
```

```python
df = pd.merge(Industry,normalized_style,left_index = True,right_index = True)
df
```

```python
Industry_name = ['cons','交通运输','休闲服务','化工','医药生物','国防军工','家用电器','建筑材料',
 '建筑装饰','房地产','有色金属','机械设备','汽车','电子','电气设备','通信',
 '采掘','银行','非银金融','食品饮料']
```

```python
rf=DataAPI.MktIborGet(ticker=u"Shibor1D",tradeDate="20200323",field=u"rate")/252  #当日无风险收益
Return = DataAPI.EquRetudGet(secID=universe,beginDate=u"20200323",endDate=u"20200323",field=u"dailyReturnNoReinv",pandas="1")
#超额收益率
Y = Return.dailyReturnNoReinv-rf.values[0,0]
```

```python
# 回归
# 生成关键向量
def calculate_ind(df):
    ratio = np.zeros(len(Industry_name +factor_list)-1)
    for i in range(1,len(Industry_name)-1):
        ratio[i] = -(weight[df[Industry_name[i]] == 1].sum()/np.sum(weight))/(weight[df[Industry_name[-1]]==1].sum()/np.sum(weight))
    return ratio

W = np.diag(np.sqrt(weight)/np.sum(np.sqrt(weight)))
X = df[Industry_name +factor_list].values
R = np.zeros((len(Industry_name +factor_list),len(Industry_name +factor_list)-1))
R[len(Industry_name)-1,:] = calculate_ind(df)
for i in range(len(Industry_name)-1):
    R[i,i] = 1
for i in range(len(Industry_name),len(Industry_name+factor_list)):
    R[i,i-1] = 1
M = X.dot(R)
S = (M.T).dot(W).dot(M)
O = R.dot(np.linalg.inv(S)).dot(M.T).dot(W)
```

```python
temp_df = pd.DataFrame(O).T
temp_df['date'] = "20200323"
temp_df['code'] = df.secID.values
temp_df.set_index(['date','code'],inplace=True)
temp_df.columns = ['市场因子']+[i+'行业因子' for i in Industry_name[1:]]+[i+'因子' for i in factor_list]
temp_df['市值加权组合'] = weight.values
temp_df
```

**验证市场加权组合和纯市场因子组合的权重相近**
