```python
"""取日期列表"""
import pandas as pd
import numpy as np
from scipy.stats import mstats
from scipy import stats
from datetime import datetime
from CAL.PyCAL import *
font.set_size(12)
#导入已经保存到lib中的alphalens
from lib.utils import *           # *号为包里所有函数的代表
from lib.plotting import *
from lib.performance import *

begin_date = datetime.strptime('2017-10-01', '%Y-%m-%d')    # 起始时间格式：2010-01-01 00:00:00        #########  修改日期为2010年1月1日--2017年12月20日  #########
end_date = datetime.strptime('2017-12-20', '%Y-%m-%d')      # 截止时间格式：2017-12-20 00:00:00

# 起始日期至截止日期之间的交易日列表
dates_list = DataAPI.TradeCalGet(exchangeCD='XSHG', beginDate=begin_date, endDate=end_date)     #交易所交易日历函数,isOpen：是否开市,prevTradeDate:前一交易日,isWeekEnd:是否当周最后交易日
dates_list = dates_list[dates_list['isOpen']==1]['calendarDate'].values.tolist()      #取开市且当天日期的值,以列表形式呈现【'2017-10-09',...】
dates_list[:5]        #取数据前五行，运行快，方便看数据格式
```

```python
"""创建因子表"""
all_universe = DynamicUniverse('A')          #证券池取A股市场
universe = all_universe.preview(dates_list[-1])            #取最后一天的股票代码列表【'002596.XSHE'】,最新日期
factor_name = 'CMRA'
for dt in dates_list:
    tmp_fac = DataAPI.MktStockFactorsOneDayGet(tradeDate=dt,secID=universe,field='secID,'+factor_name,pandas="1")      #获取多只股票历史上某一天的因子数据,field是标签，最终SecID,CMRA
    tmp_fac.columns = ['secID',dt]          #索引Index(u'secID','2017-10-09'
    if dt == dates_list[0]:
        factor_init = tmp_fac
    else:
        factor_init = pd.merge(factor_init,tmp_fac,on='secID',how='right')        #how连接方式右接,循环；on标签还是列表
factor_init = factor_init.set_index('secID').T             #以secID作为DataFrame的单独行索引,转置成列索引
factor_init.head()         #取数据前五行，运行快，方便看数据格式    行:日期,列:secID
```

```python
'''创建价格表'''
tmp_data = DataAPI.MktEqudGet(secID=universe,ticker=u"",beginDate=dates_list[0],endDate=dates_list[-1],isOpen="",field=u"secID,tradeDate,closePrice",pandas="1")
#沪深股票日行情数据,改field   secID,tradeDate,closePrice三列数据
# #########  修改日期为2010年1月1日--2017年12月20日  ########
prices_init = tmp_data.pivot_table(values='closePrice',index='tradeDate',columns='secID')       #透视表,行为交易日,列为股票代码,值为收盘价
prices_init.index = pd.to_datetime(prices_init.index.values, format='%Y-%m-%d')           #日期索引DatetimeIndex,日期字符串格式转换成标准日期格式
prices_init.head()         #取数据前五行，运行快，方便看数据格式    行:日期,列:secID   
```

```python
"""将因子表的格式转换成以日期为索引的全部股票的因子数据"""    ######### 注:只能运行一次,因为字符串格式一转换日期格式就不能再运行了
factor_init.index = pd.to_datetime(factor_init.index.values.tolist(), format='%Y-%m-%d')    #字符串转换成日期格式,不可多次运行,已转换一次
factor_init = factor_init.stack()       #stack栈,后进先出
factor_init.index.names = [u'tradeDate', 'secID']
factor_init.head(10)     #取10行数据观看数据格式,运行效率快
```

```python
''' 图1 因子值频数分布直方图 '''    #### 保存 ####

#因子数据处理：对因子数据进行截面处理：去极值，标准化
def winsorize_series(se):
    q = se.quantile([0.025, 0.975])                  #andas用方法2求得pos=1+(n-1)*p,其值返回两个分位数位置
    if isinstance(q, pd.Series) and len(q) == 2:     #判断isinstance为True或False,确保是序列且只有两个分位数位置
        se[se < q.iloc[0]] = q.iloc[0]               #1、loc——通过行标签索引行数据；  2、iloc——通过行号索引行数据；  3、ix——通过行标签或者行号索引行数据（基于loc和iloc的混合）
        se[se > q.iloc[1]] = q.iloc[1]               #头尾都替换成0.025和0.975的分位值 
    return se

def standardize_series(se):
    se_std = se.std()
    se_mean = se.mean()
    return (se - se_mean)/se_std

# factor_init = factor_df.groupby(level='tradeDate').apply(winsorize_series)      # 去极值
factor_init = factor_init.groupby(level='tradeDate').apply(standardize_series)  # 标准化            #apply应用def定义的函数
factor_init.hist(figsize=(12,6), bins=50)    #figsize：图像大小（长x宽）;    bins:直方图条数50条柱子
```
