# 工具1：股票筛选器V5.0

```python
import pandas as pd
import numpy as np
from datetime import timedelta

# 默认筛选ROIC近5年大于15%的公司
def getTickersByRatio(list_ticker = [], index = 'ROIC', level = 15, years = 5):
    # 收益率
    df = DataAPI.FdmtIndiRtnGet(ticker=list_ticker,secID="",endDate="",beginDate="",beginYear=u"",endYear=u"",reportType=u"A",field=u"",pandas="1")
    # 筛选years年内收益率指标index大于level的股票
    list_ticker = []
    for ticker in df.groupby('ticker'):
        df_temp = ticker[1][:years]
        if len(df_temp[df_temp[index]>level]) == years:
            list_ticker.append(ticker[0])
    return list_ticker

# 默认筛选近3年净现金流大于0的公司
def getTickersByCash(list_ticker = [], index = 'NChangeInCash', level = 0, years = 3):
    # 现金流
    df = DataAPI.FdmtCFAllLatestGet(ticker=list_ticker,secID=u"",reportType=u"A",endDate=u"",beginDate=u"",year="",field=u"",pandas="1")
    # 筛选years年内收益率指标index大于level的股票
    list_ticker = []
    for ticker in df.groupby('ticker'):
        df_temp = ticker[1][:years]
        if len(df_temp[df_temp[index]>level]) == years:
            list_ticker.append(ticker[0])
    return list_ticker

def getLastTradeDate():
    now = pd.datetime.today()
    start = (now-timedelta(days=10)).strftime("%Y%m%d")
    end = (now-timedelta(days=1)).strftime("%Y%m%d")
    # 交易日列表
    df_date = DataAPI.MktIdxdGet(tradeDate=u"",indexID=u"",ticker=u"000001",beginDate=start,endDate=end,exchangeCD=u"XSHE,XSHG",field=u"tradeDate",pandas="1")
    # 上一个交易日
    tradeDateLast = df_date.tradeDate.tolist()[-1]
    return tradeDateLast

# 根据PE筛选股票
def filterTickersByPE(list_ticker, PEdemand = 30):
    tradeDateLast = getLastTradeDate()
    df_PE = DataAPI.MktStockFactorsOneDayGet(tradeDate=tradeDateLast,secID=u"",ticker=list_ticker,field=u"ticker,tradeDate,pe",pandas="1")
    df_PE = df_PE[df_PE['PE']< PEdemand]
    list_ticker = df_PE.ticker.tolist()
    return list_ticker

def filterTickersByPeriod(list_ticker, seasons = 3, ratio = "grossMARgin", level = 0 ):
    df = DataAPI.FdmtIndiQGet(ticker=list_ticker,secID="",endDate="",beginDate="",beginYear=u"",endYear=u"",reportType=u"Q1,Q2,Q3,Q4",field=u"",pandas="1")

    list_ticker = []
    for ticker in df.groupby('ticker'):
        df_temp = ticker[1][:seasons+1]
        
        if ratio in ["grossMARgin"] :
            df_temp['pre' + ratio] = df_temp[ratio] 
            df_temp['pre' + ratio][df_temp.index] = df_temp[ratio][df_temp.index+1]
            df_temp[ratio] = df_temp[ratio] - df_temp['pre' + ratio]

        df_temp = df_temp.head(seasons)
        #print df_temp[['endDate','ticker','secShortName',ratio]]
        if len(df_temp[df_temp[ratio]>level]) == seasons:
            list_ticker.append(ticker[0])
    return list_ticker

# 企业价值倍数估值
def getEnterpriceValueByROIC(ticker, N = 7, marginOfSafety = 0.2):
    # 平均的ROIC
    df_roic = DataAPI.FdmtIndiRtnGet(ticker=ticker,secID="",endDate="",beginDate="",beginYear=u"",endYear=u"",reportType=u"A",field=u"",pandas="1")

    avg_roic = df_roic['ROIC'].head(5).mean()
    result_roic = df_roic['ROIC'].tolist()[0]

    if result_roic > avg_roic:
        result_roic = avg_roic

    result_roic /= 100
    
    # IC = 股东权益（TShEquity） + 有息负债
    df_IC = DataAPI.FdmtBSAllLatestGet(ticker=ticker,secID=u"",reportType=u"",endDate=u"",beginDate=u"",year="",field=u"",pandas="1")
    # 有息负债 = 短期借款(STBorr) + 1年内到期的长期负债（NCLWithin1Y） + 长期借款（LTBorr） + 应付债券（bondPayable） + 长期应付款（LTPayable）
    df_IC_result = df_IC[['TShEquity','STBorr','NCLWithin1Y','LTBorr','bondPayable','LTPayable']]
    IC = df_IC_result.loc[0].sum()
    
    # EBIT
    EBIT = result_roic * IC

    # EV
    EV = N * EBIT
    
    # 有息负债
    df_IC_debt = df_IC[['STBorr','NCLWithin1Y','LTBorr','bondPayable','LTPayable']]
    debt = np.nan_to_num(df_IC_debt.loc[0].sum())
    
    # 现金
    cash = df_IC.loc[0].cashCEquiv
    
    # 可接受的总市值
    markcap = EV - debt + cash
    
    #  总股本
    df_totalShares = DataAPI.EquShareGet(secID=u"",ticker=ticker,beginDate=u"",endDate=u"",partyID=u"",field=u"",pandas="1")
    totalShares = df_totalShares.loc[0].totalShares

    # 可接受收盘价
    price = markcap / totalShares
    
    # 安全边际打折后的收盘价
    price = round(price*(1-marginOfSafety),2)
    
    return price

# 筛选股票，默认根据ROIC筛选
def getStocks(list_ticker, index = 'ROIC', isEV = False):
    
    if isEV:
        # 初始化输出列表
        df_result = pd.DataFrame(index=list_ticker, columns=['公司名称','估值','现价','安全边际','PE','近五年平均' + index,'最新统计年度','近五年' + index + '(时间由近到远)','所属行业1','所属行业2','所属行业3'])
    else:
        # 初始化输出列表
        df_result = pd.DataFrame(index=list_ticker, columns=['公司名称','PE','近五年平均' + index,'最新统计年度','近五年' + index + '(时间由近到远)','所属行业1','所属行业2','所属行业3'])

    # 收益率
    df_ratio = DataAPI.FdmtIndiRtnGet(ticker=list_ticker,secID="",endDate="",beginDate="",beginYear=u"",endYear=u"",reportType=u"A",field=u"",pandas="1")
    

    # 申万行业分类
    industry = DataAPI.EquIndustryGet(industryVersionCD=u"",industry=u"申万行业分类",secID=u"",ticker=u"",intoDate=u"",field=u"",pandas="1")
    industry = industry.sort(columns=['outDate'], ascending=[False])
    industry = industry.drop_duplicates(['ticker'])
    
    tradeDateLast = getLastTradeDate()

    df_PE = DataAPI.MktStockFactorsOneDayGet(tradeDate=tradeDateLast,secID=u"",ticker=list_ticker,field=u"ticker,tradeDate,pe",pandas="1")

    
    for ticker in df_ratio.groupby('ticker'):
        df_temp = ticker[1][:5]
        stockcode = ticker[0]
        stockname = df_temp.secShortName.tolist()[0]
        try:
            PE = round(df_PE[df_PE['ticker']==stockcode].PE.tolist()[0],2)
        except:
            PE= np.NAN
     
        endDate = df_temp.endDate.tolist()[0]
        ratio_avg = df_temp[index].mean()

        ratio_str = ""
        ratio_list = df_temp[index].tolist()
        for ratio in ratio_list:
            ratio_str += str(round(ratio,2))
            if ratio_list.index(ratio) < len(ratio_list) - 1 :
                ratio_str += " | "
        
        # 行业信息
        industryName1 = industry[industry['ticker']==stockcode].industryName1.tolist()[0]
        industryName2 = industry[industry['ticker']==stockcode].industryName2.tolist()[0]
        industryName3 = industry[industry['ticker']==stockcode].industryName3.tolist()[0]
        
        if isEV:
            # 估值
            price = getEnterpriceValueByROIC(stockcode, N = 1/0.15, marginOfSafety = 0)
            # 现价
            closePrice = getClosePrice(stockcode)

            # 安全边际
            marginOfSafety = "%.2f%%" % ((1 - closePrice / price)*100)        
            df_result.ix[stockcode] = [stockname, price, closePrice, marginOfSafety, PE, ratio_avg, endDate, ratio_str, industryName1,industryName2,industryName3]
        else:
            df_result.ix[stockcode] = [stockname, PE, ratio_avg, endDate, ratio_str, industryName1,industryName2,industryName3]

    #df_result = df_result.sort(columns=['近五年平均' + index],ascending=[False])
    if isEV:
        df_result = df_result.sort(columns=['安全边际'],ascending=[False])
    else:
        df_result = df_result.sort(columns=['近五年平均' + index],ascending=[False])
        
    return df_result

def getClosePrice(ticker):
    price_df = DataAPI.MktEqudGet(tradeDate=u"",secID=u"",ticker=ticker,beginDate=u"",endDate=u"",isOpen="",field=u"",pandas="1")
    
    closePrice = price_df.closePrice.tolist()[-1]
    secID = price_df.secID.tolist()[-1]
    
    try:
        price_df1 = DataAPI.MktBarRTIntraDayGet(securityID=secID,startTime=u"",endTime=u"",unit=u"",pandas="1")
        closePrice = price_df1.closePrice.tolist()[-1]
    except:
        pass
    
    return closePrice

# 得到同行业股票，默认根据ROIC由高到低排序，取前十
def getStocksCorr(ticker , index = 'ROIC'):
    
    stockcode = ticker
    # 申万行业分类
    industry = DataAPI.EquIndustryGet(industryVersionCD=u"",industry=u"申万行业分类",secID=u"",ticker=u"",intoDate=u"",field=u"",pandas="1")
    industry = industry.sort(columns=['outDate'], ascending=[False])
    industry = industry.drop_duplicates(['ticker'])
    
    industryID3 = industry[industry['ticker']==stockcode].industryID3.tolist()[0]
    df_industry = industry[industry['industryID3']==industryID3].drop_duplicates(['ticker'])

    list_industryStock = df_industry.ticker.tolist()

    df_industry_ratio = DataAPI.FdmtIndiRtnGet(ticker=list_industryStock,secID="",endDate="",beginDate="",beginYear=u"",endYear=u"",reportType=u"A",field=u"",pandas="1")

    df_sort = pd.DataFrame(index=list_industryStock, columns=[index + '_AVG'])
    for stock in df_industry_ratio.groupby('ticker'):
        stockcode = stock[0]
        df_temp = stock[1][:5]
        ratio_average = df_temp[index].mean()
        df_sort.ix[stockcode] = [ratio_average]
    df_sort = df_sort.dropna()
    df_sort = df_sort.sort(columns=[index + '_AVG'], ascending=[False])
    list_corr  = df_sort.head(10).index.tolist()
    
    if ticker not in list_corr:
        list_corr.append(ticker)
  
    df_corr = DataAPI.EquGet(equTypeCD=u"A",secID=u"",ticker=list_corr,listStatusCD=u"",field=u"",pandas="1")
    df_corr['rank'] = 1
    df_corr['rank'][df_corr['ticker']==ticker] = 0
    df_corr = df_corr.sort(columns=['rank'], ascending=[True])
    df_corr['TShEquity'] = np.round(df_corr['TShEquity'] / 100000000, 2).astype(str) + "亿"
    listTitleE = ['ticker','secShortName','listDate','primeOperating','endDate','TShEquity']
    listTitleC = ['代码','简称','上市日期','主营业务范围','财务报告日期','所有者权益合计']
    df_corr = df_corr[listTitleE]
    for i in range(len(listTitleE)):
        df_corr = df_corr.rename(columns={listTitleE[i]:listTitleC[i]})
        
    return df_corr

# 主函数
def main(indexRatio = "ROIC", levelRatio = 15, yearsRatio = 5, yearsCash = 3, PEdemand = 30):
    # A股所有股票
    df_A = DataAPI.EquGet(equTypeCD=u"A",secID=u"",ticker='',listStatusCD=u"",field=u"",pandas="1")
    list_ticker = df_A['ticker'].tolist()

    # 筛选指标:收益率 ， 标准（大于）：15% ， 年限：近5年
    list_ticker = getTickersByRatio(list_ticker, index = indexRatio, level = levelRatio, years = yearsRatio)
    # 筛选指标：净现金流 ， 标准（大于）：0 ， 年限：近3年
    list_ticker = getTickersByCash(list_ticker, years = yearsCash)
    # 筛选指标：市盈率， 标准（小于）：30
    list_ticker = filterTickersByPE(list_ticker, PEdemand = PEdemand)

    # 显示股票和行业信息
    df_stocks = getStocks(list_ticker, index = indexRatio)
    return df_stocks


```
