## 二、经历
2015年的某一天下班后，夜深人静，我左手打开大智慧公式编辑器，右手翻出《日本蜡烛图》，开始编写神奇的“启明星形态”的代码。看到神奇公式筛选出了一只只底部反转的牛股，我真为自己能够结合前人宝贵经验和当前科学技术而感到“骄傲”。但我还是知道这样是不够的，而必须通过回测来验证选股结果到底如何。
由于有点VBA功底，很快我就弄好了一个模板---将大智慧的选股结果复制粘贴到excel表，VBA程序就会自动形成日期序列、每日选股、分配资金、买卖执行、计算净值。
无数次的优化后，最后得到的净值曲线其实也蛮喜人的，跟据这一套，居然在4、5月份赚了50%。
虽然六月份的行情告诉我“你赚的不过是beta而已”，我还是很怀念那一段开发山寨模型的点点滴滴：开发、回测、兴奋、发现问题、沮丧、优化、再兴奋、实盘、亏成狗。

2016年初的某一天，我打开excel表，wind的行业数据自动刷新，轻敲选股button，计算半天方才得到了今天的XX行业轮动模型的选股结果。
可是我并不满意，感觉自己对行业轮动的盘感有了更nb的认识，所以熟练按下Alt+F11,准备调整一下代码。
编程水平有限，选股、回测、业绩统计分成了三个excel模板，想到又要折腾半天，不由得一阵叹息，内心的小人在呼喊：花点心思，整合一下吧！
于是便上网搜搜有没有现成的模板，老子自己编不出代码，学习别人的代码还是很快的。

后来搜到了优矿，也终于知道原来我要做的工作已经有更专业的人做好了，而且拿来即用，还有什么比这个更让人开心呢？新学python？磨刀不误砍柴工！

## 三、研究
### 1. 多因子
作为一个在基金公司工作过的人，自然知道多因子模型对量化投资的重要性。两年的研究下来，确确实实感受到多因子模型的强大，感受到任何投资策略（甚至任何事物的发展）都是多因子模型的子集，其核心思想异曲同工。
我对广义多因子的理解是：找到驱动市场的不同核心力量（因子回归），统计每个市场力量的历史规律并预测（策略构建）其发展规律，最后通过选股和组合构建让自己的持仓顺应这种发展规律。
这句话看起来比较简单，但实际操作起来无比复杂，无数细节里有无数魔鬼。比如不同核心力量的不同指什么？统计历史规律如何选择周期维度？预测发展规律是简单线性外推还是复杂的机器学习？顺应发展规律是指绝对收益还是相对收益？
说实话我对回答这些问题也没有自信。或许现代投资组合理论已经给出了答案，比如东方证券的朱剑涛就把国外的知识很好很全面的呈现给了大家。然而这些量化模型的逻辑和绩效并不能让我满意，因为我感觉市场参与者并不是这么思考的。
最后关于alpha因子和风险因子之争，个人认为没有任何意义。正如前言所说，任何alpha因子长期来看都是beta因子，过去的非流动因子、特质波动率因子都一度被当作alpha因子；现在流行的量化基本面也不例外，市场总会有不看基本面的时候。所以，如果我们希望得到一条稳定向上的超额收益曲线，更重要还是预测这些风险因子的发展规律，当然这也是最难的。
自己做多因子（部门量化投资刚起步），数据上和基本面知识体系上实在没啥增量信息，所以更多的精力放在了因子轮动上，回测尚可，如图：

### 2. 大类资产配置
自己毕竟是做FOF投资研究的，也估计是被国外那篇资产配置解释了投资组合90%的收益的文章洗脑，在大类资产配置上做了一些研究。
个人对资产配置的理解是：分成技术面和基本面。技术面包括risk parity、BL、最小方差等资产配置模型，社区里的魏老师已经讲解的很清楚明白了。至于要不要用这些模型，又是仁者见仁了。我是不太用这些配置方法的，不过我自己的配置方法也没有提升太多sharp值。。。基本面自然是从经济增长、通胀、利率等角度出发，通过分析这些决定各类资产收益的因子来预测资产本身。背后的思想其实就是多因子，美林时钟算是基本面模型的一个特例吧。
