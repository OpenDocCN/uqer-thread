```python
from numpy import *

def hurst(main_data, region_multiplier = 2, return_points = False):
    x, y = [], []
    size = len(main_data)
    regions_vector = [region_multiplier]
    while True:
        region = regions_vector[-1] * 2
        if region > len(main_data) : break
        regions_vector.append(region)

    for local_size in regions_vector:
        data = array_split(main_data, local_size)
        mean_data = list(map(mean, data))
        mean_adjusted_series = [data[i] - mean_data[i] for i in range(local_size)]
        cumulative = list(map(cumsum, mean_adjusted_series))
        data_range = list(map(lambda x: max(x) - min(x), cumulative))
        standard_deviation = list(map(std, data))
        rescaled_range = array(data_range)/array(standard_deviation)
        mean_rescaled_range = mean(rescaled_range)
        if not isnan(mean_rescaled_range):
            fraction = size/local_size
            x.append(log2(fraction))
            y.append(log2(mean_rescaled_range))
        del data, mean_data, mean_adjusted_series, cumulative, data_range, standard_deviation, rescaled_range, mean_rescaled_range

    result = polyfit(x, y, 1)
    H = result[0]
    k = log2(result[1])
    if return_points:
        return H, k, array(x), array(y)
    else:
        return H, k
```

```python
from numpy.random import normal
from matplotlib.pyplot import *
from numpy import *

data = normal(0, 1, 6000000)
result = hurst(data, return_points = True)
H = result[0]
k = result[1]
x = result[2]
y = result[3]
plot(x, y, 'rs')
plot(x, H*x + k, 'b')
title('H = ' + str(H))
show()
```
