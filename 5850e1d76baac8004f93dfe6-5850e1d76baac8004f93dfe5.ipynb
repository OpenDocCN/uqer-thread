```python
re.sub('road$', 'rd.', s)    
```

这个$代表着 string的最后 而^则代表string的开始 这是一对相对的符号 再来两个例子

```python
##关于$的应用
print(re.sub('road$', 'rd.', 'road north 100'))## road 放在了string的最前面 并没有替换rd.
print(re.sub('road$', 'rd.', 'north 100 road123'))#以road123 结尾 并没有替换成rd.
print(re.sub('world$','wd','end of the world'))#以world结尾 替换成wd
print(re.sub('world$','wd','world of the end'))#world 放在开始 并没有提出按wd
```

```python
print(re.sub('^world','wd','world of the end'))#^是代表string的开始
print(re.sub('^road','rd.','road north 100'))
```

如果我们想要替换单词出现在最后，但是我确切的说 就是这个单词，并不是和别的单词连在一起 比如说broad，在road前面还有个b，your_world，在world前还有个your_,这并不是我想要的 
原文 at the end of the string and it was its own word (and not a part of some larger word)
在string 最后 但是并不是一个大的单词的一部分
这样的情况下 可以使用\b 意味着单词的界限在这里

```python
print(re.sub('ab*','cd','abbbbbbb'))# 在这个例子中 只要出现ab。
                                    #无论a后面跟着多少个b 都会直接替换 而单个的a也会被替换
print(re.sub('ab*','cd','ab'))
print(re.sub('ab*','cd','a'))
print(re.sub('ab*','cd','ae'))
print('-----------------------------------')

print(re.sub('abc*','xyz','abcdefg'))
print(re.sub('abc*','xyz','abcccdefg'))
print(re.sub('abc*','xyz','aQbbccccccccccccccdefg'))
print(re.sub('abc*','xyz','abbccdefg'))##这个例子非常有趣abc 可以 abcccc可以 
#但是a 不可以 然后 ab却可以
#其实和 ab* 中的a 应该是一个道理
```

```python
print('1:'+re.sub('ab+','cd','abbbbbbb'))## + 和*类似 唯一的区别就是不会再和单单一个a匹配
print('2:'+re.sub('ab+','cd','ab'))
print('3:'+re.sub('ab+','cd','a'))
print('4:'+re.sub('ab+','cd','ae'))
#从第三行的代码可以看出  ab+一定要和ab 或者abbbbb 匹配，单单一个a无法匹配到

print('-----------------------------------')
print('1:'+re.sub('abc+','xyz','abcdefg'))
print('2:'+re.sub('abc+','xyz','abcccdefg'))
print('3:'+re.sub('abc+','xyz','aQbbccccccccccccccdefg'))
print('4:'+re.sub('abc+','xyz','abbccdefg'))
#abc+则是一定要和abc匹配
#第三行出现了a
#第四行代码中出现了ab 但是这两行代码都没有与abc+匹配到
```

```python
print(re.sub('\\bRoad$', 'rd.', 'north 100 Road'))
print(re.sub('\\bRoad$', 'rd.', 'north 100 broad'))
print(re.sub('\\bworld$','wd','end of the world'))
print(re.sub('\\bworld$','wd','end of the your_world'))
```

```python
print('1:'+re.sub('<.*>','nv','sometimes <tooNaive> and 2 young'))
print('2:'+re.sub('<.*>','nv','sometimes <tooNaive> and 2 <young>'))
print('3:'+re.sub('<.*>','nv','sometimes <tooNaive and 2 young>'))
print('4:'+re.sub('<.*>','nv','<sometimes> <tooNaive> and 2 young'))
print('5:'+re.sub('<.*?>','nv','sometimes <tooNaive> and 2 young'))
print('6:'+re.sub('<.*?>','nv','sometimes <tooNaive> and 2 <young>'))
##<.*> "is matched against <a> b <c>, it will match the entire string, and not just <a>. Adding ? after the qualifier makes it perform the match in non-greedy or minimal fashion; as few characters as possible will be matched. Using the RE <.*?> will match only <a>."

#官方文档说 <.*> 是一种贪婪搜索 不仅仅是会搜索<a>会搜索 整个<a>b<c> 从第2行代码 中可以看到。
# <tooNaive> 和<young> 中间的 and 2 一起被替换成了nv 这个就是<a>b<c> 的结构
#但是加了？以后就不同了，？ 让他尽可能少的匹配，使用的是非贪婪搜索， 5-6行的代码 也能够清晰的看出来
```

```python
print(re.sub('\(.*?\)','nv','sometimes (tooNaive) and 2 <young>'))
print(re.sub('(.*?)','nv','sometimes (tooNaive) and 2 <young>'))
print(re.sub('\[.*?\]','nv','sometimes <tooNaive> and 2 [young]'))
print(re.sub('[.*?]','nv','sometimes <tooNaive> and 2 [young]'))
print(re.sub('\[.*?\]','nv','sometimes [tooNaive] and 2 [young]'))
print(re.sub('\G.*?\G','nv','sometimes GtooNaiveG and 2 [young]'))
print(re.sub('\?.*?\?','nv','sometimes ?tooNaive? and 2 [young]'))
print(re.sub('\*.*?\*','nv','sometimes *tooNaive* and 2 [young]'))
#提醒一下 这个如果是想要 替换掉()的和[]里的 或者说 任意被一对字符包括在内的 就像是我6,7,8行代码的例子一样 分别是G,?,*这一对里面的内容 需要添加\ \，不然就会想第2行代码和第4行代码一样出现问题。
```

```python
print('{m}')
print(re.sub('a{2}','b','aa'))
print(re.sub('a{2}','b','aaa'))
print(re.sub('a{2}','b','temp a c aa'))
#{m} 这个{m}代表着 {m}之前的string重复多少次
#{m,n} 而{m,n}则代表着重复几次倒几次，{2,3}就是重复2，3次。7 8的行的代码就看到了aa和aaa都能被替代了
print('------------------------------------')#手动分割
print('{m,n}')
print(re.sub('a{2,3}','b','aG'))
print(re.sub('a{2,3}','b','aaG'))
print(re.sub('a{2,3}','b','aaaG'))
print(re.sub('a{2,3}','b','aaaaG'))
print('------------------------------------')#手动分割
#{m,n}? 加了问号 以后 使得搜索变得不贪婪，尽量满足m 而不是n 13行print 结果 4个a 结果是搜索出来2组aa 于是替换成两个b
print(re.sub('a{2,4}?','b','aaaaG'))
print(re.sub('a{2,4}','b','aaaG'))### 3个a 满足{2,4}2到4 但是还没有到4 所以直接替换成b
print(re.sub('a{2,4}','b','aaaaaG'))#5个a 我觉得这个挺奇怪的，他判断为直接把4个a替换成b，但是不是 两组aa，再落单一个a
print(re.sub('a{2,4}','b','aaaaaaG'))#6个a往上全部没有判断为aa 为一组。从例子来看，都是应该以{2,4}。 这边我也有些不清楚，可能只是数量比较少的时候会用到{m,n}?这样吧
print(re.sub('a{2,4}','b','aaaaaaaG'))
print(re.sub('a{2,4}','b','aaaaaaaaG'))
print(re.sub('a{2,3}','b','aG')) #一个a并不满足 所以并没有替换 这个比较简单

```

```python
#[] 中括号的用法， 其实一开始就已经有几个例子了
print(re.sub('[amk]','G','amktoo_young'))#e.g. [amk] will match 'a', 'm', or 'k'. 官方文档说的会和a m k任意匹配，
                                         #把amk 全部替换了
print(re.sub('[amk]','G','atmoko_young'))#  我把a m k 拆开来看看
#[0-5][0-9] will match all the two-digits numbers from 00 to 59 [0-5][0-9]这个会匹配00-59所有的
print(re.sub('[0-5][0-9]','G','49too_young'))
print(re.sub('[0-5][0-9]','G','60too_young'))#60就不匹配咯
print(re.sub('[0-5][0-9]','G','0too_young'))
print(re.sub('[0-5][0-9][1-2]','G','591too_young'))#试试看三位数行不行0-592，591可以
print(re.sub('[0-5][0-9][1-2]','G','600too_young'))#试试看三位数行不行0-592，600不可以 4位数也可以 以此类推 我已经试过了
```

```python
#[^5] will match any character except '5' ^这个我觉得挺有用的 匹配所有不是5的字符串
print(re.sub('[^5]','G','49too_young')) #不过他是单个字符一个一个检查 然后一个字符一个字符替换 成G
print(re.sub('[^naive]','young','toonaive'))#所以这边把非naive 也就是too转换成了3个young
#其实一般的用法可能就是直接替换成空格
print(re.sub('[^naive]','','toonaive'))#这样非naive的too 就直接没了
```

```python
print('1:'+re.sub('young|naive','','toonaive'))# |代表着 or 在我的string里搜索young 或者 naive
print('2:'+re.sub('young|naive','','toonaive_and_too_young'))#young 和naive 都没了
print('3:'+re.sub('[|]', 'G','too|naive'))#如果只是想替换掉| 这个符号，需要加上中括号[|] 类似得，如果想搜寻* + ？这些特                                            殊字符就加上中括号
                                          #或者 加上\ 变成\| 3,5行代码应该比较清楚
print('4:'+re.sub('\|', 'G','too|naive'))
print('3:'+re.sub('[*?+]', '','*too?naive+'))#字符串里面的* ？+ 全都没有了
```
