```python
import pandas as pd
import numpy as np
from dateutil.parser import parse
from datetime import date, timedelta
import datetime
import cPickle as pickle
import os
import time
from enum import Enum
import sys

dir_path = "esg"
if not os.path.isdir(dir_path):
    os.mkdir(dir_path)
    print(u"创建数据文件夹:{0}".format(dir_path))
else:
    print(u"数据文件夹:{0}已存在".format(dir_path))

class TimeDateFormat(Enum):
    """
    时间格式, 枚举类型
    """
    YMDHYPHEN = '%Y-%m-%d'
    YMD = '%Y%m%d'
    YMDHMSHYPHEN = '%Y-%m-%d %H:%M:%S'
    YMDHMS = '%Y:%m:%d %H:%M:%S'
    Y = '%Y'
    HM = '%H:%M'
    
class TimeDateUtil:
    @staticmethod
    def convert_str_to_date(src_str, src_format):
        """
        将timestamp格式日期，转换为date
        """
        try:
            d = datetime.datetime.strptime(src_str, src_format)
        except ValueError:
            d = None
        except TypeError:
            d = None
        return d

    @staticmethod
    def convert_date_to_str(src_date, tgt_format):
        """
        将date格式日期，转换为timestamp
        """
        return src_date.strftime(tgt_format)

    @staticmethod
    def convert_format(src_str, src_format, tgt_format):
        """
        转换日期格式
        :param src_str: 源格式日期
        :param src_format: 源格式
        :param tgt_format: 目标格式
        :return d: 转换格式后的日期
        """
        src_date = TimeDateUtil.convert_str_to_date(src_str, src_format)
        if src_date is None:
            return None
        return TimeDateUtil.convert_date_to_str(src_date, tgt_format)
    
    @staticmethod
    def get_previous_date_str(src_str, src_format, tgt_format, n):
        """
        指定日期前N天日期
        """
        last_days = TimeDateUtil.convert_str_to_date(src_str, src_format) - timedelta(n)
        return last_days.strftime(tgt_format)
    
    @staticmethod
    def reset_time_format(df, src_format, tgt_format, column_name='TRADE_DATE'):
        """
        转换trade_date格式为"yyyy"
        :return:
        """
        new_time_index = list(map(lambda x: TimeDateUtil.convert_format(str(x), src_format, tgt_format), df[column_name].values))
        df[column_name] = pd.Series(new_time_index, index=df.index)
        return df
    
    @staticmethod
    def get_time_delta(src_str, tgt_str, src_format, tgt_format, unit='DAY'):
        """
        两日期间时间差
        """
        last_days = TimeDateUtil.convert_str_to_date(src_str, src_format) - TimeDateUtil.convert_str_to_date(tgt_str, tgt_format)
        if unit == 'DAY':
            return last_days.days
        elif unit == 'SEC':
            return last_days.seconds
        return last_days.days

def load_pickle(pickle_name, ori_obj):
    """
    加载之前处理好的缓存文件
    :param pickle_name: 存储文件名
    :param ori_obj: 变量默认值
    :return valid, ori_obj: 是否之前存在, 读取后的变量值
    """
    if not os.path.exists(pickle_name):
        return False, ori_obj
    else:
        with open(pickle_name, 'rb') as handle:
            ori_obj = pickle.load(handle)
    return True, ori_obj

def save_pickle(pickle_name, ori_obj):
    """
    存储变量为缓存文件
    :param pickle_name: 存储文件名
    :param ori_obj: 变量
    """
    with open(pickle_name, 'wb') as handle:
        pickle.dump(ori_obj, handle, protocol=pickle.HIGHEST_PROTOCOL)
    return

def dataframe_apply_int(df, column):
    """
    将str格式列，转换为int
    """
    df[column] = df[column].apply(int)
    return df

def filter_security(df, trade_cal, benchmark):
    """
    按投资域筛选因子
    统计覆盖率
    """
    new_df = []
    coverage = []
    bm_ticker = '000300' if benchmark == 'hs300' else '000906' if benchmark == 'zz800' else '000985'
    for idx, (_, rows) in enumerate(trade_cal.iterrows()):
        td = rows['calendarDate']
        td_no_hyphen = rows['calendarDate'].replace('-', '')
        cons_tickers = DataAPI.IdxConsGet(secID=u"",ticker=bm_ticker,isNew=u"",intoDate=td_no_hyphen,field=u"consTickerSymbol",pandas="1")
        cons_tickers_set = cons_tickers['consTickerSymbol'].values.tolist()
        ticker_in_condition = df['ticker'].apply(lambda x: x in cons_tickers_set)
        sub_df = df[(ticker_in_condition) & (df['tradeDate'] == td)].copy(deep=True)
        new_df.append(sub_df)
        coverage.append([td, len(sub_df.index) / float(len(cons_tickers_set))])
        if idx % 10 == 0:
            print(u'[FilterSecurity][{0}/{1}] benchmark:{2} done.'.format(idx, len(trade_cal.index), benchmark))
    new_df = pd.concat(new_df, ignore_index=True)
    coverage_df = pd.DataFrame(data=coverage, columns=['tradeDate', 'coverage'])
    return new_df, coverage_df
```

```python
def get_fiscal_df(read_begin_date, fiscal_type='A', fiscal_period='12'):
    """
    取财务数据的真实发布日期
    """
    equ_fiscal_df = DataAPI.FdmtISGet(ticker=u"",secID=u"",reportType=fiscal_type,endDate=u"",beginDate=u"",publishDateEnd=u"",
                  publishDateBegin=read_begin_date,endDateRep="",beginDateRep="",beginYear="",endYear="",fiscalPeriod=fiscal_period,
                  field=u"ticker,publishDate,endDate,actPubtime,operateProfit,tRevenue",pandas="1")
    equ_fiscal_df = equ_fiscal_df.sort_values(by=['ticker', 'endDate', 'publishDate'], ascending=[True, True, True]).drop_duplicates(
        subset=['ticker', 'endDate'], keep='first')
    return equ_fiscal_df
        
def compute_count_factor(df, trade_cal, factor_name):
    """
    计数类因子计算
    """
    df = df.copy(deep=True)
    for _, rows in trade_cal.iterrows():
        td = rows['calendarDate']
        valid = (df['beginDate'] <= td) & ((df['endDate'] > td) | (df['endDate'].isnull()))
        df[td] = 0.0
        df.loc[valid, td] = 1.0

    tds = [rows['calendarDate'] for _, rows in trade_cal.iterrows()]
    factor_df = df.groupby('ticker')[tds].sum().reset_index().set_index('ticker').stack().reset_index().rename(
        columns={'level_1': 'tradeDate', 0: factor_name})
    return factor_df

def compute_count_in_window_factor(df, trade_cal, factor_name, window):
    """
    时间窗内计数类因子计算
    """
    df = df.copy(deep=True)
    for _, rows in trade_cal.iterrows():
        td = rows['calendarDate']
        three_years_ago = TimeDateUtil.get_previous_date_str(td, TimeDateFormat.YMDHYPHEN.value, TimeDateFormat.YMDHYPHEN.value, window*365)
        valid = (df['beginDate'] <= td) & (df['beginDate'] > three_years_ago)
        df[td] = 0.0
        df.loc[valid, td] = 1.0

    tds = [rows['calendarDate'] for _, rows in trade_cal.iterrows()]
    factor_df = df.groupby('ticker')[tds].sum().reset_index().set_index('ticker').stack().reset_index().rename(
        columns={'level_1': 'tradeDate', 0: factor_name})
    return factor_df

def compute_sum_in_window_factor(df, trade_cal, col_name, factor_name, window):
    """
    时间窗内加和类因子计算
    """
    df = df.copy(deep=True)
    for _, rows in trade_cal.iterrows():
        td = rows['calendarDate']
        three_years_ago = TimeDateUtil.get_previous_date_str(td, TimeDateFormat.YMDHYPHEN.value, TimeDateFormat.YMDHYPHEN.value, window*365)
        valid = (df['beginDate'] <= td) & (df['beginDate'] > three_years_ago)
        df[td] = 0.0
        df.loc[valid, td] = df.loc[valid, col_name]

    tds = [rows['calendarDate'] for _, rows in trade_cal.iterrows()]
    factor_df = df.groupby('ticker')[tds].sum().reset_index().set_index('ticker').stack().reset_index().rename(
        columns={'level_1': 'tradeDate', 0: factor_name})
    return factor_df

def compute_ratio_factor(df_num, df_den, factor_name_num, factor_name_den, factor_name):
    """
    比例类因子计算
    """
    df_num = df_num.copy(deep=True).set_index(['ticker', 'tradeDate'])
    df_den = df_den.copy(deep=True).set_index(['ticker', 'tradeDate'])
    factor_df = (df_num[factor_name_num] / df_den[factor_name_den]).reset_index().rename(columns={0: factor_name})
    return factor_df

def compute_fiscal_count_factor(df, trade_cal, column_name, factor_name, key=['ticker']):
    """
    根据财务日历，计数类因子计算
    """
    df = df.copy(deep=True)
    for _, rows in trade_cal.iterrows():
        td = rows['calendarDate']
        valid = df[df['publishDate'] <= td].groupby(key).tail(1).index
        df[td] = 0.0
        df.loc[valid, td] = df.loc[valid, column_name]

    tds = [rows['calendarDate'] for _, rows in trade_cal.iterrows()]
    factor_df = df.groupby('ticker')[tds].sum().reset_index().set_index('ticker').stack().reset_index().rename(
        columns={'level_1': 'tradeDate', 0: factor_name})
    return factor_df

def compute_managers_info_education_factor(df, trade_cal, factor_name):
    """
    董事会:独立董事平均教育程度因子
    """
    df = df.copy(deep=True)
    df['bachelor'] = df['education'].str.contains('.*本科.*')
    df['master'] = df['education'].str.contains('.*(硕士|研究生).*')
    df['doctor'] = df['education'].str.contains('.*博士.*')
    for _, rows in trade_cal.iterrows():
        td = rows['calendarDate']
        valid = (df['beginDate'] <= td) & ((df['endDate'] > td) | (df['endDate'].isnull()))
        df[td] = 0.0
        df.loc[(valid & df['bachelor']), td] = 1.0 / 3.0
        df.loc[(valid & df['master']), td] = 2.0 / 3.0
        df.loc[(valid & df['doctor']), td] = 1.0

    tds = [rows['calendarDate'] for _, rows in trade_cal.iterrows()]
    factor_df = df.groupby('ticker')[tds].mean().reset_index().set_index('ticker').stack().reset_index().rename(
        columns={'level_1': 'tradeDate', 0: factor_name})
    return factor_df

def compute_managers_info_age_factor(df, trade_cal, factor_name):
    """
    董事会:独立董事平均年龄因子
    """
    df = df.copy(deep=True)
    for _, rows in trade_cal.iterrows():
        td = rows['calendarDate']
        sixty_years_ago = TimeDateUtil.get_previous_date_str(td, TimeDateFormat.YMDHYPHEN.value, TimeDateFormat.Y.value, 60*365)
        valid = (df['beginDate'] <= td) & ((df['endDate'] > td) | (df['endDate'].isnull()))
        df[td] = 0.0
        df.loc[(valid & (df['birthday'] < sixty_years_ago)), td] = 1.0

    tds = [rows['calendarDate'] for _, rows in trade_cal.iterrows()]
    factor_df = df.groupby('ticker')[tds].mean().reset_index().set_index('ticker').stack().reset_index().rename(
        columns={'level_1': 'tradeDate', 0: factor_name})
    return factor_df

def merge_factors(factors):
    """
    合并多个因子df为一个df
    """
    final_df = pd.DataFrame()
    for idx, factor_df in enumerate(factors):
        final_df = factor_df if idx == 0 else final_df
        if idx > 0:
            final_df = final_df.merge(factor_df, left_on=['ticker', 'tradeDate'], right_on=['ticker', 'tradeDate'], how='outer')
    final_df = final_df.replace([np.inf, -np.inf], np.nan)
    final_df = final_df.fillna(0.0)
    return final_df

def compute_board_factors(read_begin_date, begin_date, end_date):
    """
    董事会因子
    ---合理性
    1) 董事会规模(BOARD_SIZE,-),EquManagersGet
    2) 董事长CEO是否是同一人
    3) 董监高平均学历[学士、硕士、博士分别计1/3、2/3、1分](IND_DIR_AVG_EDU,+),EquManagersInfoGet
    4) 繁忙董事监事比例(IND_DIR_PART_RATIO,-),EquManagersGet,EquManagementConcurrentlyGet
    5) 监事会会议次数(SUPERVISOR_NUM,+),EquMeetInforGet
    6) 委员会个数(COMMITTEE_NUM,+),EquCommitteeGet
    ---稳定性
    7) 3年内CEO变动（不含退休、任期届满）次数(CEO_CHANGE_NUM,-),EquManageChgGet
    8) 3年内董事长变动（不含退休、任期届满）次数(CHAIRMAN_CHANGE_NUM,-),EquManageChgGet
    ---独立性
    9) 董事会独立董事比例(IND_DIR_NUM_RATIO,+),EquManagersGet
    10) 独立董事财务专家比例(IND_DIR_FIN_RATIO,+),EquManagersGet
    11) 独立董事法律专家比例(IND_DIR_LAW_RATIO,+),EquManagersGet
    12) 独立董事是否出席所有董事会会议(IND_DIR_ATTEND_RATIO,+),EquIndependentDirectorsVotingGet
    13) 独立董事高于60岁比例(IND_DIR_ABOVE_SIXTY,-),EquManagersInfoGet
    14) 独立董事工作地点是否与上市公司一致
    15) 独立董事是否有繁忙董事,EquManagementConcurrentlyGet
    """
    future_three_years = TimeDateUtil.get_previous_date_str(start_date, TimeDateFormat.YMD.value, TimeDateFormat.YMDHYPHEN.value, -3 * 365)
    trade_cal = DataAPI.TradeCalGet(exchangeCD=u"XSHG", beginDate=begin_date, endDate=end_date,field=u"", pandas="1")
    trade_cal = trade_cal[(trade_cal['isMonthEnd'] == 1) & (trade_cal['isOpen'] == 1)]
    universe = DataAPI.SecIDGet(partyID=u"", ticker=u"", cnSpell=u"", assetClass=u"E", field=u"", pandas="1")
    universe = universe[(universe['exchangeCD'] == 'XSHE') | (universe['exchangeCD'] == 'XSHG')]
    gap = int(len(universe['ticker'].values) / 5)
    t0 = time.time()
    # 1) 董事会任职
    equ_managers_df = DataAPI.EquManagersGet(secID=u"",ticker=u"",beginDate=read_begin_date,endDate=u"",field=u"ticker,managerName,managerType,position,beginDate,endDate",pandas="1")
    manager_type_df = dataframe_apply_int(DataAPI.SysCodeGet(codeTypeID=u"20045",valueCD=u"",field=u"valueCD,valueName",pandas="1"), 'valueCD')
    equ_managers_df = equ_managers_df.merge(manager_type_df, left_on=['managerType'], right_on=['valueCD'], how='left').rename(columns={'valueName': 'managerTypeCN'}).drop(['valueCD'], axis=1)
    factor_board_size_df = compute_count_factor(equ_managers_df, trade_cal, 'BOARD_SIZE')
    factor_ind_dir_ratio_df = compute_ratio_factor(compute_count_factor(equ_managers_df[equ_managers_df['position'].str.contains('.*独立董事.*')], trade_cal, 'IND_DIR_NUM'), 
                                                  factor_board_size_df, 'IND_DIR_NUM', 'BOARD_SIZE', 'IND_DIR_NUM_RATIO')
    factor_ind_dir_fin_ratio_df = compute_ratio_factor(compute_count_factor(equ_managers_df[equ_managers_df['position'].str.contains('.*财务.*')], trade_cal, 'IND_DIR_FIN_NUM'), 
                                                      factor_board_size_df, 'IND_DIR_FIN_NUM', 'BOARD_SIZE', 'IND_DIR_FIN_RATIO')
    factor_ind_dir_law_ratio_df = compute_ratio_factor(compute_count_factor(equ_managers_df[equ_managers_df['position'].str.contains('.*法律.*')], trade_cal, 'IND_DIR_LAW_NUM'), 
                                                      factor_board_size_df, 'IND_DIR_LAW_NUM', 'BOARD_SIZE', 'IND_DIR_LAW_RATIO')
    t1 = time.time()
    # 2) 董事会基本信息
    equ_managers_info_df = DataAPI.EquManagersInfoGet(secID=u"",ticker=u"",field=u"ticker,managerName,birthday,education",pandas="1")
    equ_managers_info_df = equ_managers_df.merge(equ_managers_info_df, left_on=['ticker', 'managerName'], right_on=['ticker', 'managerName'], how='left')
    factor_ind_dir_avg_edu_df = compute_managers_info_education_factor(equ_managers_info_df, trade_cal, 'IND_DIR_AVG_EDU')
    factor_ind_dir_above_sixty_df = compute_managers_info_age_factor(equ_managers_info_df, trade_cal, 'IND_DIR_ABOVE_SIXTY')
    t2 = time.time()
    # 3) 监事会会议信息
    equ_fiscal_df = get_fiscal_df(read_begin_date, fiscal_type='A', fiscal_period="12")
    equ_meet_info_df = DataAPI.EquMeetInforGet(secID=u"",ticker=u"",beginDate=read_begin_date,endDate=u"",
                                               field=u"ticker,endDate,directorNum,supervisorNum,shareholderNum",pandas="1").fillna(0.0)
    equ_meet_info_df = equ_meet_info_df.merge(equ_fiscal_df, left_on=['ticker', 'endDate'], right_on=['ticker', 'endDate'], how='left')
    factor_supervisor_num_df = compute_fiscal_count_factor(equ_meet_info_df, trade_cal, 'supervisorNum', 'SUPERVISOR_NUM', key=['ticker'])
    t3 = time.time()
    # 4) 委员会
    equ_committee_df = DataAPI.EquCommitteeGet(
        secID=u"",ticker=u"",beginDate=read_begin_date,endDate=u"",field=u"ticker,publishDate,comName,accessionDate,dimissionDate",pandas="1").rename(
        columns={'accessionDate': 'beginDate', 'dimissionDate': 'endDate'})
    equ_committee_df['endDate'] = equ_committee_df['endDate'].fillna(future_three_years)
    equ_committee_df = (equ_committee_df.groupby(by=['ticker', 'publishDate', 'comName'])['beginDate'].min().reset_index()).merge(
        equ_committee_df.groupby(by=['ticker', 'publishDate', 'comName'])['endDate'].max().reset_index(), 
        left_on=['ticker', 'publishDate', 'comName'], right_on=['ticker', 'publishDate', 'comName'])
    factor_committee_df = compute_count_factor(equ_committee_df, trade_cal, 'COMMITTEE_NUM')
    t4 = time.time()
    # 5) 高管变动
    equ_managers_chg_df = DataAPI.EquManageChgGet(secID=u"",ticker=u"",beginDate=read_begin_date,endDate=u"",
                                                  field=u"ticker,publishDate,changeDate,changePosition,actionType,quitType",pandas="1").rename(columns={'publishDate': 'beginDate'})
    position_chg_type_df = DataAPI.SysCodeGet(codeTypeID=u"20090", valueCD=u"", field=u"valueCD,valueName", pandas="1")
    action_type_df = DataAPI.SysCodeGet(codeTypeID=u"20097", valueCD=u"", field=u"valueCD,valueName", pandas="1")
    quit_type_df = DataAPI.SysCodeGet(codeTypeID=u"20093", valueCD=u"", field=u"valueCD,valueName", pandas="1")
    equ_managers_chg_df = equ_managers_chg_df.merge(position_chg_type_df, left_on=['changePosition'], right_on=['valueCD'], how='left').rename(columns={'valueName': 'changePositionCN'}).drop(['valueCD'], axis=1)
    equ_managers_chg_df = equ_managers_chg_df.merge(action_type_df, left_on=['actionType'], right_on=['valueCD'], how='left').rename(columns={'valueName': 'actionTypeCN'}).drop(['valueCD'], axis=1)
    equ_managers_chg_df = equ_managers_chg_df.merge(quit_type_df, left_on=['quitType'], right_on=['valueCD'], how='left').rename(columns={'valueName': 'quitTypeCN'}).drop(['valueCD'], axis=1).dropna()
    equ_managers_chg_df = equ_managers_chg_df[~equ_managers_chg_df['quitTypeCN'].str.contains('(退休|任职届满)')]
    factor_ceo_chg_df = compute_count_in_window_factor(equ_managers_chg_df[equ_managers_chg_df['changePositionCN'] == '总经理'], trade_cal, 'CEO_CHANGE_NUM', window=3)
    factor_chairman_chg_df = compute_count_in_window_factor(equ_managers_chg_df[equ_managers_chg_df['changePositionCN'] == '董事长'], trade_cal, 'CHAIRMAN_CHANGE_NUM', window=3)
    t5 = time.time()
    # 6) 独立董事
    equ_id_att_df = DataAPI.EquIndependentDirectorsVotingGet(secID=u"",ticker=u"",beginDate=read_begin_date,
                                                             endDate=u"",field=u"ticker,endDate,name,attendanceNum,entrustedAttNum,absNum",pandas="1")
    equ_id_att_df['attendRatio'] = (equ_id_att_df['attendanceNum'] - equ_id_att_df['absNum']) / equ_id_att_df['attendanceNum']
    equ_id_att_df = equ_id_att_df.merge(equ_fiscal_df, left_on=['ticker', 'endDate'], right_on=['ticker', 'endDate'], how='left')
    equ_id_att_df = equ_id_att_df.groupby(by=['ticker', 'endDate', 'publishDate'])['attendRatio'].mean().reset_index()
    factor_id_att_df = compute_fiscal_count_factor(equ_id_att_df, trade_cal, 'attendRatio', 'IND_DIR_ATTEND_RATIO', key=['ticker'])
    t6 = time.time()
    # 7) 繁忙董事
    equ_managers_con_df = []
    for i in range(5):
        equ_managers_con_df.append(DataAPI.EquManagementConcurrentlyGet(secID=u"",ticker=universe['ticker'].values.tolist()[i*gap:(i+1)*gap],
                                                                        beginDate=read_begin_date,endDate=u"",field=u"ticker,endDate,name,partUnitName",pandas="1"))
    equ_managers_con_df = pd.concat(equ_managers_con_df, ignore_index=True)
    equ_managers_con_df = equ_managers_con_df.merge(equ_fiscal_df, left_on=['ticker', 'endDate'], right_on=['ticker', 'endDate'], how='left')
    equ_managers_con_df = equ_managers_con_df.groupby(by=['ticker', 'endDate', 'publishDate', 'name'])['partUnitName'].count().reset_index().rename(columns={'partUnitName': 'partNum'})
    equ_managers_con_df = equ_managers_con_df[equ_managers_con_df['partNum'] >= 4].groupby(by=['ticker', 'endDate', 'publishDate'])['name'].count().reset_index().rename(columns={'name': 'partNum'})
    factor_ind_dir_part_ratio_df = compute_ratio_factor(compute_fiscal_count_factor(equ_managers_con_df, trade_cal, 'partNum', 'ID_DIR_PART_NUM_CON'),
                                                        factor_board_size_df, 'ID_DIR_PART_NUM_CON', 'BOARD_SIZE', 'IND_DIR_PART_RATIO')
    t7 = time.time()
    print('[PreloadData] board factor computed, cost: {0:.2f}'.format(t7 - t0))
    return [factor_board_size_df, factor_ind_dir_ratio_df, factor_ind_dir_fin_ratio_df, factor_ind_dir_law_ratio_df,
            factor_ind_dir_avg_edu_df, factor_ind_dir_above_sixty_df, factor_supervisor_num_df, factor_committee_df, factor_ceo_chg_df,
            factor_chairman_chg_df, factor_id_att_df, factor_ind_dir_part_ratio_df]
```

```python
def get_qfiis():
    """
    自己维护的qfii机构名单
    """
    qfiis = set(["挪威中央银行","BILL AND MELINDA GATES FOUNDATION TRUST","UBS AG","邓普顿投资顾问有限公司","长城-中行-景顺资产管理有限公司-景顺中国系列基金",
                 "NOMURA SECURITIES CO., LTD","东方汇理银行","CITIGROUP GLOBAL MARKETS LIMITED","ING BANK N.V.",
                 "SCBHK A/C BBH S/A VANGUARD EMERGING MARKETS STOCK INDEX FUND","申银万国-农行-BNP PARIBAS","摩根士丹利中国A股基金",
                 "瑞士信贷(香港)有限公司","DEUTSCHE BANK AKTIENGESELLSCHAFT","国泰君安-建行-香港上海汇丰银行有限公司","海通-交行-日兴资产管理有限公司-日兴AM中国人民币A股母基金",
                 "英国保诚资产管理(香港)有限公司","哥伦比亚大学","中银国际-中行-法国爱德蒙得洛希尔银行","JPMORGAN CHASE BANK,NATIONAL ASSOCIATION",
                 "高华-汇丰-GOLDMAN, SACHS AND CO.","柏瑞投资有限责任公司","GOVERNMENT OF SINGAPORE INV. CORP.- A/C \"C\"","YALE UNIVERSITY",
                 "MORGAN STANLEY AND CO. INTERNATIONAL PLC","资本国际公司","海通-中行-富通银行","MERRILL LYNCH INTERNATIONAL",
                 "POWER CORPORATION OF CANADA","ABU DHABI INVESTMENT AUTHORITY","The Bank of Tokyo-Mitsubishi UFJ Ltd.","JF资产管理有限公司-JF中国先驱Ａ股基金",
                 "第一生命保险株式会社","新加坡华侨银行有限公司(QFII)","通用电气资产管理公司-GEAM信托基金中国A股基金","未来资产基金管理公司-未来资产中国A股股票型母投资信托",
                 "WANG SHANG KEE AND/OR CHIN WAN LAN","淡马锡","瑞银环球资产管理(新加坡)-瑞银卢森堡机构SICAVII中国Ａ股","高盛国际资产管理公司-高盛国际中国基金",
                 "国泰君安-中行-渣打银行(香港)有限公司","荷宝基金管理公司-客户资金","百达资产管理有限公司-客户资金","魁北克储蓄投资集团","野村资产管理株式会社-野村中国投资基金",
                 "安石投资管理有限公司-安石大中华基金","KB资产运用-KB中国大陆基金","首域投资管理(英国)有限公司-首域中国A股基金","群益证券投资信托股份有限公司-客户资金",
                 "亢简资产管理公司-客户资金","国泰证券投资信托股份有限公司-客户资金","韩华资产运用株式会社-韩华中国本土母基金","法国欧菲资产管理公司-欧菲汉朝基金"])
    return qfiis

def get_daily_market(begin_date, end_date):
    """
    计算日线因子
    """
    trade_cal = DataAPI.TradeCalGet(exchangeCD=u"XSHG", beginDate=begin_date, endDate=end_date,field=u"", pandas="1")
    trade_cal = trade_cal[(trade_cal['isMonthEnd'] == 1) & (trade_cal['isOpen'] == 1)]
    data = []
    for _, rows in trade_cal.iterrows():
        daily_df = DataAPI.MktEqudGet(secID=u"",ticker=u"",tradeDate=rows['calendarDate'],beginDate=u"",endDate=u"",isOpen="",field=u"ticker,tradeDate,closePrice,marketValue",pandas="1")
        daily_df['marketVol'] = daily_df['marketValue'] / daily_df['closePrice']
        data.append(daily_df)
    daily_df = pd.concat(data, ignore_index=True)
    return daily_df

def compute_hk_exchange_factor(df, trade_cal, factor_name):
    """
    计算个股是否在港股同上市的因子
    """
    df = df.copy(deep=True)
    for _, rows in trade_cal.iterrows():
        td = rows['calendarDate']
        valid = (df['listDate'] <= td) & (df['listDateHK'] <= td)
        df[td] = 0.0
        df.loc[valid, td] = 1.0

    tds = [rows['calendarDate'] for _, rows in trade_cal.iterrows()]
    factor_df = df.groupby('ticker')[tds].max().reset_index().set_index('ticker').stack().reset_index().rename(
        columns={'level_1': 'tradeDate', 0: factor_name})
    return factor_df
    
def compute_ownership_factors(read_begin_date, begin_date, end_date):
    """
    股东大会因子
    ---控制权与所有权
    1) 股权集中度指标（前8大股东股票比例,TOP_SHARE_HOLDER_PCT,-）,EquShTenGet
    2) 公司是否有实际控制人(ACTUAL_CONTROLLER_HOLDER_PCT,-),EquActualControllerGet
    3) 两权分离度（控制权/所有权,CONTROL_HOLDER_PCT,-）,EquActualControllerGet/EquShTenGet
    ---内部持股
    4) 监事持股比例(SUPERVISOR_HOLDER_PCT,+),EquMsChangesGet
    5) 高管持股比例(MANAGER_HOLDER_PCT,+),EquMsChangesGet
    6) 管理层（除高管外）持股比例(NORMAL_MANAGER_HOLDER_PCT,+),EquMsChangesGet
    ---外部持股
    7) 基金持股比例(FUND_HOLD_RATIO_PCT,+),FundHoldingsGet
    8) 社保基金持股比例(SSF_SHARE_HOLDER_PCT,+),EquFloatShTenGet
    9) QFII持股比例(QFII_SHARE_HOLDER_PCT,+),EquFloatShTenGet
    10) 非金融类上市公司持股比例
    11) 是否发行H股(WITH_HK_SHARE,+),SecIDGet
    """
    trade_cal = DataAPI.TradeCalGet(exchangeCD=u"XSHG", beginDate=begin_date, endDate=end_date,field=u"", pandas="1")
    trade_cal = trade_cal[(trade_cal['isMonthEnd'] == 1) & (trade_cal['isOpen'] == 1)]
    equ_fiscal_df = get_fiscal_df(read_begin_date, fiscal_type=u"", fiscal_period=u"")
    daily_df = get_daily_market(begin_date, end_date)
    universe = DataAPI.SecIDGet(partyID=u"", ticker=u"", cnSpell=u"", assetClass=u"E", field=u"", pandas="1")
    universe = universe[(universe['exchangeCD'] == 'XSHE') | (universe['exchangeCD'] == 'XSHG')]
    gap = int(len(universe['ticker'].values) / 5)
    t0 = time.time()
    # 1) 前十大股东
    equ_sh_ten_df = []
    for i in range(5):
        equ_sh_ten_df.append(DataAPI.EquShTenGet(secID=u"",ticker=universe['ticker'].values.tolist()[i*gap:(i+1)*gap],beginDate=read_begin_date,
                                                 endDate=u"",field=u"ticker,endDate,shNum,shName,holdVol,holdPct",pandas="1"))
    equ_sh_ten_df = pd.concat(equ_sh_ten_df, ignore_index=True)
    equ_sh_ten_df = equ_sh_ten_df.merge(equ_fiscal_df, left_on=['ticker', 'endDate'], right_on=['ticker', 'endDate'], how='left')
    equ_sh_ten_df = (equ_sh_ten_df[equ_sh_ten_df['shNum'] <= 8]).groupby(by=['ticker', 'endDate', 'publishDate'])['holdPct'].sum().reset_index()
    factor_top_share_ratio_df = compute_fiscal_count_factor(equ_sh_ten_df, trade_cal, 'holdPct', 'TOP_SHARE_HOLDER_PCT', key=['ticker'])
    t1 = time.time()
    # 2) 前十大流通股东
    qfiis = get_qfiis()
    equ_sh_ten_float_df = []
    for i in range(5):
        equ_sh_ten_float_df.append(DataAPI.EquFloatShTenGet(secID=u"",ticker=universe['ticker'].values.tolist()[i*gap:(i+1)*gap],
                                 beginDate=read_begin_date,endDate=u"",field=u"ticker,endDate,publishDate,shNum,shName,holdVol,holdPct",pandas="1"))
    equ_sh_ten_float_df = pd.concat(equ_sh_ten_float_df, ignore_index=True)
    factor_ssf_share_ratio_df = compute_fiscal_count_factor(((equ_sh_ten_float_df[equ_sh_ten_float_df['shName'].str.contains('.*社保.*')]).groupby(
        by=['ticker', 'endDate', 'publishDate'])['holdPct'].sum().reset_index()), trade_cal, 'holdPct', 'SSF_SHARE_HOLDER_PCT', key=['ticker'])
    factor_qfii_share_ratio_df = compute_fiscal_count_factor(((equ_sh_ten_float_df[equ_sh_ten_float_df['shName'].apply(lambda x: x in qfiis)]).groupby(
        by=['ticker', 'endDate', 'publishDate'])['holdPct'].sum().reset_index()), trade_cal, 'holdPct', 'QFII_SHARE_HOLDER_PCT', key=['ticker'])
    t2 = time.time()
    # 3) 实际控制人
    equ_actual_controller_df = DataAPI.EquActualControllerGet(secID=u"",ticker=u"",beginDate=read_begin_date,endDate=u"",field=u"ticker,endDate,actContro,controlRadio",pandas="1")
    equ_actual_controller_df = equ_actual_controller_df.merge(equ_fiscal_df, left_on=['ticker', 'endDate'], right_on=['ticker', 'endDate'], how='left')
    equ_actual_controller_df = equ_actual_controller_df.groupby(by=['ticker', 'endDate', 'publishDate'])['controlRadio'].sum().reset_index()
    factor_actual_controller_ratio_df = compute_fiscal_count_factor(equ_actual_controller_df, trade_cal, 'controlRadio', 'ACTUAL_CONTROLLER_HOLDER_PCT', key=['ticker'])
    factor_control_hold_ratio_df = compute_ratio_factor(factor_actual_controller_ratio_df, factor_top_share_ratio_df, 'ACTUAL_CONTROLLER_HOLDER_PCT', 'TOP_SHARE_HOLDER_PCT', 'CONTROL_HOLDER_PCT')
    t3 = time.time()
    # 4) 高管持股
    equ_ms_change_df = DataAPI.EquMsChangesGet(secID=u"",ticker=u"",beginDate=read_begin_date,endDate=u"",
        field=u"ticker,managerName,position,shareChanges,shareHoldAf,changeProp,changeDate,fillDate",pandas="1").rename(columns={'fillDate': 'publishDate'})
    equ_ms_change_df = equ_ms_change_df.dropna(subset=['shareHoldAf', 'position'])
    tmp_market_vol_df = compute_fiscal_count_factor(daily_df.rename(columns={'tradeDate': 'publishDate'}), trade_cal, 'marketVol', 'MARKET_VOL', key=['ticker'])
    factor_supervisor_ratio_df = compute_ratio_factor(compute_fiscal_count_factor(equ_ms_change_df[equ_ms_change_df['position'].str.contains('.*监事.*')], trade_cal, 'shareHoldAf',
        'SUPERVISOR_HOLDER_PCT', key=['ticker', 'managerName']), tmp_market_vol_df, 'SUPERVISOR_HOLDER_PCT', 'MARKET_VOL', 'SUPERVISOR_HOLDER_PCT')
    factor_manager_ratio_df = compute_ratio_factor(compute_fiscal_count_factor(equ_ms_change_df[equ_ms_change_df['position'].str.contains('.*(高管|高级管理).*')], trade_cal, 'shareHoldAf',
        'MANAGER_HOLDER_PCT', key=['ticker', 'managerName']), tmp_market_vol_df, 'MANAGER_HOLDER_PCT', 'MARKET_VOL', 'MANAGER_HOLDER_PCT')
    factor_normal_manager_ratio_df = compute_ratio_factor(compute_fiscal_count_factor(equ_ms_change_df[~equ_ms_change_df['position'].str.contains('.*(高管|高级管理).*')], trade_cal,
        'shareHoldAf','NORMAL_MANAGER_HOLDER_PCT', key=['ticker', 'managerName']), tmp_market_vol_df, 'NORMAL_MANAGER_HOLDER_PCT', 'MARKET_VOL', 'NORMAL_MANAGER_HOLDER_PCT')
    t4 = time.time()
    # 5) 基金持股
    equ_fund_hold_df = []
    for i in range(5):
        equ_fund_hold_df.append(DataAPI.FundHoldingsGet(secID=u"",ticker=universe['ticker'].values.tolist()[i*gap:(i+1)*gap],reportDate=u"",
            beginDate=read_begin_date,endDate=u"",secType="E",field=u"ticker,holdingTicker,holdVolume,marketValue,reportDate,publishDate",pandas="1")
                                .rename(columns={'ticker': 'fundTicker', 'holdingTicker': 'ticker', 'marketValue': 'holdValue'}))
    equ_fund_hold_df = pd.concat(equ_fund_hold_df, ignore_index=True)
    equ_fund_hold_df = equ_fund_hold_df.merge(daily_df, left_on=['ticker', 'reportDate'], right_on=['ticker', 'tradeDate'])
    equ_fund_hold_df['holdPct'] = equ_fund_hold_df['holdValue'] * 100.0 / equ_fund_hold_df['marketValue']
    factor_fund_hold_ratio_df = compute_fiscal_count_factor(equ_fund_hold_df, trade_cal, 'holdPct', 'FUND_HOLD_RATIO_PCT', key=['ticker', 'fundTicker'])
    t5 = time.time()
    # 6) 发行H股
    security_df = DataAPI.SecIDGet(partyID=u"",ticker=u"",cnSpell=u"",assetClass=u"E",field=u"ticker,exchangeCD,listDate,partyID",pandas="1")
    a_share_df = security_df[security_df['exchangeCD'].apply(lambda x: x in ['XSHE', 'XSHG'])]
    hk_share_df = security_df[security_df['exchangeCD'].apply(lambda x: x in ['XHKG'])].rename(columns={'ticker': 'tickerHK', 'exchangeCD': 'exchangeCDHK', 'listDate': 'listDateHK'})
    a_share_df = a_share_df.merge(hk_share_df, left_on=['partyID'], right_on=['partyID'], how='left')
    factor_with_hk_share_df = compute_hk_exchange_factor(a_share_df, trade_cal, 'WITH_HK_SHARE')
    t6 = time.time()
    print('[PreloadData] onwership factor computed, cost: {0:.2f}'.format(t6 - t0))
    return [factor_top_share_ratio_df, factor_ssf_share_ratio_df, factor_qfii_share_ratio_df, factor_actual_controller_ratio_df,
            factor_control_hold_ratio_df, factor_supervisor_ratio_df, factor_manager_ratio_df, factor_normal_manager_ratio_df,
            factor_fund_hold_ratio_df, factor_with_hk_share_df]
```

```python
def compute_st_factor(trade_cal, factor_name):
    """
    上市公司是否ST的因子
    """
    data = []
    for _, rows in trade_cal.iterrows():
        td = rows['calendarDate']
        equ_sec_st_df = DataAPI.SecSTGet(beginDate=td,endDate=td,secID=u"",ticker=u"",field=u"ticker,tradeDate,STflg",pandas="1")
        equ_sec_st_df['STflg'] = 1.0
        data.append(equ_sec_st_df)
    factor_df = pd.concat(data, ignore_index=True).rename(columns={'STflg': factor_name})
    return factor_df

def compute_finance_governance_factors(read_begin_date, begin_date, end_date):
    """
    财务治理因子
    ---财报审计
    1) 财报审计出具标准无保留意见(AFS,+),ReportByTickerGet
    2) 3年内审计事务所变动次数(AF_CHANGE,-),ReportByTickerGet
    3) 审计师得分（来自四大+1 分，来自境外+1）(BIG_FOUR,+),ReportByTickerGet
    ---内部控制(还没自生产，所以简单判断)
    4) 内控报告是否有效(INTERNAL_CONTROL,+),ReportByTickerGet
    5) 内部控制是否有问题
    6) 内控审计报告出具标准无保留意见
    ---业绩预告
    7) 财报公布日调整次数
    8) 业绩快报、业绩预告是否准确(EE_PROFIT_ERROR,-),FdmtEeGet/FdmtEfGet
    9) 是否为ST,SecSTGet(ST_FLAG,-)
    """
    trade_cal = DataAPI.TradeCalGet(exchangeCD=u"XSHG", beginDate=begin_date, endDate=end_date,field=u"", pandas="1")
    trade_cal = trade_cal[(trade_cal['isMonthEnd'] == 1) & (trade_cal['isOpen'] == 1)]
    equ_fiscal_df = get_fiscal_df(read_begin_date, fiscal_type=u"", fiscal_period=u"")
    universe = DataAPI.SecIDGet(partyID=u"", ticker=u"", cnSpell=u"", assetClass=u"E", field=u"", pandas="1")
    universe = universe[(universe['exchangeCD'] == 'XSHE') | (universe['exchangeCD'] == 'XSHG')]
    t0 = time.time()
    # 1) 公告类
    equ_report_df = DataAPI.ReportByTickerGet(ticker=universe['ticker'].values.tolist(),beginDate=read_begin_date,endDate=u"",field=u"publishDate,tiker,title",pandas="1").rename(columns={'publishDate': 'beginDate', 'tiker': 'ticker'})
    equ_report_df = equ_report_df.dropna(subset=['title'])
    factor_afs_df = compute_count_in_window_factor(equ_report_df[equ_report_df['title'].str.contains('.*财报审计.*')], trade_cal, 'AFS', window=1)
    factor_af_change_df = compute_count_in_window_factor(equ_report_df[equ_report_df['title'].str.contains('.*更换.*事务所.*')], trade_cal, 'AF_CHANGE', window=3)
    factor_big_four_df = compute_count_in_window_factor(equ_report_df[equ_report_df['title'].str.contains('.*(普华永道|德勤|毕马威|安永).*')], trade_cal, 'BIG_FOUR', window=3)
    factor_internal_control_df = compute_count_in_window_factor(equ_report_df[equ_report_df['title'].str.contains('.*(内部控制|内控).*')], trade_cal, 'INTERNAL_CONTROL', window=1)
    t1 = time.time()
    # 2) 业绩快报、预告
    equ_fdmt_ee_df = DataAPI.FdmtEeGet(ticker=u"",secID=u"",reportType=u"",endDate=u"",beginDate=read_begin_date,publishDateEnd=u"",publishDateBegin=u"",
        field=u"ticker,publishDate,endDate,operateProfit",pandas="1").rename(columns={'publishDate': 'publishDatePred', 'operateProfit': 'operateProfitPred'})
    equ_fdmt_ee_df = equ_fdmt_ee_df.merge(equ_fiscal_df, left_on=['ticker', 'endDate'], right_on=['ticker', 'endDate'], how='left')
    equ_fdmt_ee_df['publishDate'] = pd.concat([equ_fdmt_ee_df['publishDate'], equ_fdmt_ee_df['publishDatePred']], axis=1).max(axis=1)
    equ_fdmt_ee_df['predError'] = abs(equ_fdmt_ee_df['operateProfit'] - equ_fdmt_ee_df['operateProfitPred']) / equ_fdmt_ee_df['tRevenue']
    factor_fdmt_ee_error_df = compute_fiscal_count_factor(equ_fdmt_ee_df, trade_cal, 'predError', 'EE_PROFIT_ERROR', key=['ticker'])
    t2 = time.time()
    # 3) 是否ST
    factor_st_df = compute_st_factor(trade_cal, 'ST_FLAG')
    t3 = time.time()
    print('[PreloadData] finance governance factor computed, cost: {0:.2f}'.format(t3 - t0))
    return [factor_afs_df, factor_af_change_df, factor_big_four_df, factor_internal_control_df, factor_fdmt_ee_error_df, factor_st_df]
```

```python
def compute_pay_factors(read_begin_date, begin_date, end_date):
    """
    薪酬激励因子
    ---薪酬
    1) log(董事年薪前三名)(DIR_ANN_REWARD,+),EquExecsHoldingsGet
    2) log(高管年薪前三名)(MANAGER_ANN_REWARD,+),EquExecsHoldingsGet
    ---激励
    3) 员工持股计划/总股本
    4) 高管员工持股计划/总股本,EquMsChangesGet
    5) 股权激励权益/总股本(SHARE_EXCIT_RATIO,+),EquSharesExcitGet
    6) 高管激励权益/总股本,还没对外开放API,EquManageExcitGet
    """
    trade_cal = DataAPI.TradeCalGet(exchangeCD=u"XSHG", beginDate=begin_date, endDate=end_date,field=u"", pandas="1")
    trade_cal = trade_cal[(trade_cal['isMonthEnd'] == 1) & (trade_cal['isOpen'] == 1)]
    t0 = time.time()
    # 1) 高管持股薪酬计划
    equ_execs_holding_df = DataAPI.EquExecsHoldingsGet(secID=u"",ticker=u"",beginDate=read_begin_date,endDate=u"",field=u"ticker,publishDate,name,position,annReward",pandas="1")
    equ_execs_holding_df = equ_execs_holding_df.dropna(subset=['position', 'annReward'])
    factor_dir_ann_reward_df = compute_fiscal_count_factor(equ_execs_holding_df[equ_execs_holding_df['position'].str.contains('.*董事.*')], trade_cal, 'annReward', 'DIR_ANN_REWARD', key=['ticker', 'name'])
    factor_dir_ann_reward_df['DIR_ANN_REWARD'] = np.log(factor_dir_ann_reward_df['DIR_ANN_REWARD'] + sys.float_info.min)
    factor_managers_ann_reward_df = compute_fiscal_count_factor(equ_execs_holding_df[~equ_execs_holding_df['position'].str.contains('.*董事.*')], trade_cal, 'annReward', 'MANAGER_ANN_REWARD', key=['ticker', 'name'])
    factor_managers_ann_reward_df['MANAGER_ANN_REWARD'] = np.log(factor_managers_ann_reward_df['MANAGER_ANN_REWARD'] + sys.float_info.min)
    t1 = time.time()
    # 2) 股权激励
    equ_shares_excit_df = DataAPI.EquSharesExcitGet(secID=u"",ticker=u"",beginDate=read_begin_date,endDate=u"",field=u"ticker,excitSharesPct,shcPublishDate",pandas="1").rename(columns={'shcPublishDate': 'publishDate'})
    factor_shares_excit_df = compute_fiscal_count_factor(equ_shares_excit_df, trade_cal, 'excitSharesPct', 'SHARE_EXCIT_RATIO', key=['ticker'])
    t2 = time.time()
    print('[PreloadData] pay factor computed, cost: {0:.2f}'.format(t2 - t0))
    return [factor_dir_ann_reward_df, factor_managers_ann_reward_df, factor_shares_excit_df]
```

```python
def compute_behavior_factors(read_begin_date, begin_date, end_date):
    """
    治理行为及外部监管因子
    ---治理行为
    1) 董监高质押股票比例(DIR_SUP_PLEDGE_RATIO,-),EquPledgeFreezeGet
    2) 管理费用率(ADMIN_EXP_RATIO,-),FdmtISGet
    3) 资产周转率(ASSET_TURNOVER,+),FdmtISGet,FdmtBSGet
    4) 大股东占款率(OTHER_RECEIVE_RATIO,-)（其他应收款/总资产）,FdmtBSGet
    5) 关联人进货销货率(RELATED_SALES_RATIO,-)（（关联人采购额+关联人销售额）/净资产）,EquRelatedTransactionGet,FdmtBSGet
    6) 关联人融资保证比率(GUARANTEE_RATIO,-)（（向关联方提供资金余额+关联担保余额）/净资产）,EquAccumulatedGuaranteeGet,FdmtBSGet
    7) 3年内并购次数(MERGER_ACQUISITION_NUM,-)（关联交易）,EquRestructuringGet
    8) 3年内资产剥离次数(MERGER_ACQUISITION_SPLIT_NUM,-)（关联交易）,EquRestructuringGet
    9) 股权再融资情况(ALLOT_SHARES_RATIO,-)（是否增长超过5%）,EquAllotGet,EquSpoGet,BondConvIssueNewGet
    10) 商誉/净资产(GOODWILL_RATIO,-),FdmtBSGet
    11) 重大资产重组失败数(MERGER_ACQUISITION_FAIL_NUM,-),EquRestructuringGet
    12) 3年内重大资产重组停牌交易日数,SecHaltGet
    ---外部评价
    13) 近1年来违规次数(COMP_ILLEGAL_COUNT,-),EquCompIllegalGet
    14) 近半年来诉讼仲裁涉案金额数/营业收入(SUIT_AMOUNT_RATIO,-),FdmtISGet
    15) 冻结股票比例(FREEZE_RATIO,-),EquPledgeFreezeGet
    16) 投资者调研次数(INVEST_ACTIVITY_NUM,+),EquIsActivityGet
    17) 调研机构、媒体数(INVEST_ACTIVITY_ORG_NUM,+),EquIsParticipantQaGet
    """
    trade_cal = DataAPI.TradeCalGet(exchangeCD=u"XSHG", beginDate=begin_date, endDate=end_date,field=u"", pandas="1")
    trade_cal = trade_cal[(trade_cal['isMonthEnd'] == 1) & (trade_cal['isOpen'] == 1)]
    daily_df = get_daily_market(begin_date, end_date)
    universe = DataAPI.SecIDGet(partyID=u"", ticker=u"", cnSpell=u"", assetClass=u"E", field=u"", pandas="1")
    universe = universe[(universe['exchangeCD'] == 'XSHE') | (universe['exchangeCD'] == 'XSHG')]
    t0 = time.time()
    # 1) 股票质押
    equ_pledge_freeze_df = DataAPI.EquPledgeFreezeGet(secID=u"",ticker=u"",pubBeginDate=u"",pubEndDate=u"",beginDate=read_begin_date,endDate=u"",thawBeginDate=u"",thawEndDate=u"",
                               field=u"ticker,publishDate,shName,involvesum,typeselect",pandas="1")
    select_type_df = dataframe_apply_int(DataAPI.SysCodeGet(codeTypeID=u"20053",valueCD=u"",field=u"valueCD,valueName",pandas="1"), 'valueCD')
    equ_pledge_freeze_df = equ_pledge_freeze_df.merge(select_type_df, left_on=['typeselect'], right_on=['valueCD'], how='left').rename(columns={'valueName': 'typeselectCN'}).drop(['valueCD'], axis=1)
    equ_managers_df = DataAPI.EquManagersGet(secID=u"",ticker=u"",beginDate=read_begin_date,endDate=u"",field=u"ticker,managerName,position,beginDate,endDate",pandas="1")
    equ_pledge_freeze_df = equ_pledge_freeze_df.merge(equ_managers_df, left_on=['ticker', 'shName'], right_on=['ticker', 'managerName'], how='left')
    equ_pledge_freeze_df = equ_pledge_freeze_df[equ_pledge_freeze_df.apply(lambda x: x['publishDate'] > x['beginDate'] and (x['endDate'] is None or x['publishDate'] < x['endDate']), axis=1)].drop_duplicates(subset=['ticker', 'publishDate', 'shName'], keep='last')
    factor_dir_sup_pledge_ratio_df = compute_ratio_factor(compute_fiscal_count_factor(equ_pledge_freeze_df[(equ_pledge_freeze_df['position'].str.contains('.*(董事|监事).*')) & (equ_pledge_freeze_df['typeselectCN'] == '股权质押')], trade_cal, 'involvesum', 'DIR_SUP_PLEDGE_RATIO', key=['ticker', 'shName']), daily_df, 'DIR_SUP_PLEDGE_RATIO', 'marketVol', 'DIR_SUP_PLEDGE_RATIO')
    factor_freeze_ratio_df = compute_ratio_factor(compute_fiscal_count_factor(equ_pledge_freeze_df[equ_pledge_freeze_df['typeselectCN'] == '股权冻结'], trade_cal, 'involvesum', 'FREEZE_RATIO', key=['ticker', 'shName']), daily_df, 'FREEZE_RATIO', 'marketVol', 'FREEZE_RATIO')
    t1 = time.time()
    # 2) 利润表、资产负债表
    equ_fdmt_is_df = DataAPI.FdmtISGet(ticker=u"",secID=u"",reportType=u"",endDate=u"",beginDate=read_begin_date,publishDateEnd=u"",publishDateBegin=u"",endDateRep="",beginDateRep="",
                      beginYear="",endYear="",fiscalPeriod="",field=u"ticker,publishDate,endDate,tRevenue,adminExp",pandas="1")
    equ_fdmt_bs_df = DataAPI.FdmtBSGet(ticker=u"",secID=u"",reportType=u"",endDate=u"",beginDate=read_begin_date,publishDateEnd=u"",publishDateBegin=u"",endDateRep="",beginDateRep="",
                      beginYear="",endYear="",fiscalPeriod="",field=u"ticker,publishDate,endDate,TAssets,TLiab,othReceiv,goodwill,TShEquity",pandas="1")
    equ_fdmt_df = equ_fdmt_is_df.merge(equ_fdmt_bs_df, left_on=['ticker', 'publishDate', 'endDate'], right_on=['ticker', 'publishDate', 'endDate'], how='left').sort_values(
        by=['ticker', 'endDate', 'publishDate'], ascending=[True, True, True]).drop_duplicates(subset=['ticker', 'endDate'], keep='first')
    equ_fdmt_df['ADMIN_EXP_RATIO'] = equ_fdmt_df['adminExp'] / equ_fdmt_df['tRevenue']
    equ_fdmt_df['ASSET_TURNOVER'] = equ_fdmt_df['tRevenue'] / equ_fdmt_df['TAssets']
    equ_fdmt_df['OTHER_RECEIVE_RATIO'] = equ_fdmt_df['othReceiv'] / equ_fdmt_df['TAssets']
    equ_fdmt_df['GOODWILL_RATIO'] = equ_fdmt_df['goodwill'] / equ_fdmt_df['TAssets']
    factor_admin_exp_ratio_df = compute_fiscal_count_factor(equ_fdmt_df, trade_cal, 'ADMIN_EXP_RATIO', 'ADMIN_EXP_RATIO', key=['ticker'])
    factor_asset_turnover_df = compute_fiscal_count_factor(equ_fdmt_df, trade_cal, 'ASSET_TURNOVER', 'ASSET_TURNOVER', key=['ticker'])
    factor_other_receive_ratio_df = compute_fiscal_count_factor(equ_fdmt_df, trade_cal, 'OTHER_RECEIVE_RATIO', 'OTHER_RECEIVE_RATIO', key=['ticker'])
    factor_goodwill_ratio_df = compute_fiscal_count_factor(equ_fdmt_df, trade_cal, 'GOODWILL_RATIO', 'GOODWILL_RATIO', key=['ticker'])
    t2 = time.time()
    # 3) 关联交易
    equ_related_trans_df = DataAPI.EquRelatedTransactionGet(secID=u"",ticker=u"",beginDate=read_begin_date,endDate=u"",
        field=u"ticker,publishDate,relatedPartyName,transactionAmount",pandas="1").rename(columns={'publishDate': 'publishDateRelTrans'})
    equ_related_trans_df = equ_related_trans_df.merge(equ_fdmt_df, left_on=['ticker'], right_on=['ticker'], how='left')
    equ_related_trans_df = equ_related_trans_df.dropna(subset=['publishDateRelTrans', 'publishDate'])
    equ_related_trans_df['diffDays'] = equ_related_trans_df.apply(lambda x: TimeDateUtil.get_time_delta(
        x['publishDate'], x['publishDateRelTrans'], TimeDateFormat.YMDHYPHEN.value, TimeDateFormat.YMDHYPHEN.value, 'DAY'), axis=1)
    equ_related_trans_df = equ_related_trans_df[equ_related_trans_df['diffDays'] >= 0.0].sort_values(by=['ticker', 'publishDateRelTrans', 'diffDays'], ascending=[True, True, False])
    equ_related_trans_df = equ_related_trans_df.loc[equ_related_trans_df.groupby(by=['ticker', 'publishDateRelTrans', 'relatedPartyName']).tail(1).index]
    equ_related_trans_df['RELATED_SALES_RATIO'] = equ_related_trans_df['transactionAmount'] * 10000.0 / equ_related_trans_df['TShEquity']
    factor_related_trans_df = compute_fiscal_count_factor(equ_related_trans_df, trade_cal, 'RELATED_SALES_RATIO', 'RELATED_SALES_RATIO',
                                                          key=['ticker', 'publishDateRelTrans', 'relatedPartyName'])
    equ_related_guarantee_df = DataAPI.EquAccumulatedGuaranteeGet(secID=u"",ticker=u"",pubBeginDate=u"",pubEndDate=u"",beginDate=read_begin_date,endDate=u"",
        field=u"ticker,endDate,guaranteeTotalAmount,guaranteeAmountRelated",pandas="1").rename(columns={'endDate': 'endDateRel'}).fillna(0.0)
    equ_related_guarantee_df = equ_related_guarantee_df.merge(equ_fdmt_df, left_on=['ticker', 'endDateRel'], right_on=['ticker', 'endDate'], how='left')
    equ_related_guarantee_df['GUARANTEE_RATIO'] = (equ_related_guarantee_df['guaranteeTotalAmount'] + equ_related_guarantee_df['guaranteeAmountRelated']) / \
        equ_related_guarantee_df['TShEquity']
    factor_guarantee_ratio_df = compute_fiscal_count_factor(equ_related_guarantee_df, trade_cal, 'GUARANTEE_RATIO', 'GUARANTEE_RATIO', key=['ticker'])
    t3 = time.time()
    # 4) 并购重组
    equ_restruct_df = DataAPI.EquRestructuringGet(secID=u"",ticker=u"",beginDate=read_begin_date,endDate=u"",
                                                  field=u"ticker,publishDate,iniPublishDate,isRelevance,restructuringType,isSucceed",pandas="1")
    equ_restruct_df = equ_restruct_df[equ_restruct_df['isRelevance'] == 1]
    equ_restruct_df['beginDate'] = pd.concat([equ_restruct_df['publishDate'], equ_restruct_df['iniPublishDate']], axis=1).min(axis=1)
    factor_ma_num_df = compute_count_in_window_factor(equ_restruct_df, trade_cal, 'MERGER_ACQUISITION_NUM', window=3)
    factor_ma_split_num_df = compute_count_in_window_factor(equ_restruct_df[equ_restruct_df['restructuringType'] == '2'], trade_cal, 'MERGER_ACQUISITION_SPLIT_NUM', window=3)
    factor_ma_fail_num_df = compute_count_in_window_factor(equ_restruct_df[equ_restruct_df['isSucceed'] == 0], trade_cal, 'MERGER_ACQUISITION_FAIL_NUM', window=3)
    t4 = time.time()
    # 5) 再融资
    equ_allot_df = DataAPI.EquAllotGet(secID=u"",ticker=u"",isAllotment=u"1",beginDate=read_begin_date,endDate=u"",field=u"ticker,publishDate,allotShares",pandas="1")
    equ_spo_df = DataAPI.EquSpoGet(secID=u"",ticker=u"",beginListPublishDate=u"",endListPublishDate=u"",beginRecordDate=u"",endRecordDate=u"",beginDate=read_begin_date,
                      endDate=u"",beginShcPublishDate=u"",endShcPublishDate=u"",beginListDate=u"",endListDate=u"",spoTypeCD=u"",
                      field=u"planPublishDate,ticker,issueShares",pandas="1").rename(columns={'planPublishDate': 'publishDateSPO'}).dropna(subset=['issueShares'])
    equ_allot_df = equ_allot_df.merge(equ_spo_df, left_on=['ticker'], right_on=['ticker'], how='outer')
    equ_allot_df.loc[equ_allot_df['publishDateSPO'].isnull(), 'publishDateSPO'] = equ_allot_df.loc[equ_allot_df['publishDateSPO'].isnull(), 'publishDate']
    equ_allot_df.loc[equ_allot_df['publishDate'].isnull(), 'publishDate'] = equ_allot_df.loc[equ_allot_df['publishDate'].isnull(), 'publishDateSPO']
    equ_allot_df = equ_allot_df.fillna(0.0)
    equ_allot_df['diffDays'] = equ_allot_df.apply(lambda x: TimeDateUtil.get_time_delta(
        x['publishDate'], x['publishDateSPO'], TimeDateFormat.YMDHYPHEN.value, TimeDateFormat.YMDHYPHEN.value, 'DAY'), axis=1)
    equ_allot_df = equ_allot_df[equ_allot_df['diffDays'] >= 0.0].sort_values(by=['ticker', 'publishDateSPO', 'diffDays'], ascending=[True, True, False])
    equ_allot_df = equ_allot_df.loc[equ_allot_df.groupby(by=['ticker', 'publishDateSPO']).tail(1).index]
    equ_allot_df['ALLOT_SHARES'] = equ_allot_df['allotShares'] + equ_allot_df['issueShares']
    factor_allot_ratio_df = compute_ratio_factor(compute_fiscal_count_factor(equ_allot_df, trade_cal, 'ALLOT_SHARES', 'ALLOT_SHARES', key=['ticker']),
                                                 daily_df, 'ALLOT_SHARES', 'marketVol', 'ALLOT_SHARES_RATIO')
    t5 = time.time()
    # 6) 违规/仲裁
    equ_comp_illegal_df = DataAPI.EquCompIllegalGet(secID=u"",ticker=u"",beginDate=read_begin_date,endDate=u"",field=u"ticker,publishDate",pandas="1").rename(columns={'publishDate': 'beginDate'})
    equ_comp_illegal_df = equ_comp_illegal_df.drop_duplicates(subset=['ticker', 'beginDate'], keep='first')
    factor_comp_illegal_cnt_df = compute_count_in_window_factor(equ_comp_illegal_df, trade_cal, 'COMP_ILLEGAL_COUNT', window=1)
    equ_law_suit_df = DataAPI.EquLawSuitsGet(secID=u"",ticker=universe['ticker'].values.tolist(),partyID=u"",beginDate=read_begin_date,endDate=u"",lawsuitType=u"",
                                             field=u"ticker,publishDate,suitAmount",pandas="1").rename(columns={'publishDate': 'beginDate'})
    tmp_fdmt_revenue_df = compute_fiscal_count_factor(equ_fdmt_df, trade_cal, 'tRevenue', 'tRevenue', key=['ticker'])
    factor_law_suit_df = compute_ratio_factor(compute_sum_in_window_factor(equ_law_suit_df, trade_cal, 'suitAmount', 'SUIT_AMOUNT_SUM', window=0.5),
                                              tmp_fdmt_revenue_df, 'SUIT_AMOUNT_SUM', 'tRevenue', 'SUIT_AMOUNT_RATIO')
    t6 = time.time()
    # 7) 投资者调研
    equ_activity_df = DataAPI.EquIsActivityGet(secID=u"",ticker=u"",beginDate=read_begin_date,endDate=u"",field=u"ticker,publishDate,partyNum",pandas="1").rename(columns={'publishDate': 'beginDate'})
    factor_invest_num_df = compute_count_in_window_factor(equ_activity_df, trade_cal, 'INVEST_ACTIVITY_NUM', window=1)
    factor_invest_org_num_df = compute_sum_in_window_factor(equ_activity_df, trade_cal, 'partyNum', 'INVEST_ACTIVITY_ORG_NUM', window=1)
    t7 = time.time()
    print('[PreloadData] behavior factor computed, cost: {0:.2f}'.format(t7 - t0))
    return [factor_dir_sup_pledge_ratio_df, factor_freeze_ratio_df, factor_admin_exp_ratio_df, factor_asset_turnover_df, factor_other_receive_ratio_df,
            factor_goodwill_ratio_df, factor_related_trans_df, factor_guarantee_ratio_df, factor_ma_num_df, factor_ma_split_num_df, factor_ma_fail_num_df,
            factor_allot_ratio_df, factor_comp_illegal_cnt_df, factor_law_suit_df, factor_invest_num_df, factor_invest_org_num_df]
```

```python
# 单因子测试框架
import pandas as pd
import numpy as np
import datetime
from quartz.universe import set_universe
from dateutil.parser import parse
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter
# CAL.PyCAL中包含font
from CAL.PyCAL import *
cal = Calendar('China.SSE')
import scipy.stats as st
from scipy.stats.mstats import gmean
from random import randint
font.set_size(15)

class TradeDateProcess():
    """
    交易日历相关处理
    """
    def __init__(self):
        self.trade_date = None

    def update_trade_date(self):
        """
        更新获取交易日历
        """
        trade_cal = DataAPI.TradeCalGet(exchangeCD=u"XSHG",beginDate=u"",endDate=u"",field=u"isOpen,calendarDate",pandas="1")
        trade_cal = trade_cal[trade_cal['isOpen']==1]
        self.trade_date = trade_cal['calendarDate'].map(lambda x:parse(x)).tolist()
    
    def _get_date_list(self,start_date,end_date,freq=None,point=None):
        """
        可以得到指定频率交易日期列表
        :param start_date: 起始日期
        :param end_date: 终止日期
        :param freq: 频率, 日、周、月
        :param point: 取哪一天作为调仓日
        """
        calendar_date = DataAPI.TradeCalGet(exchangeCD=u"XSHG",beginDate=start_date,endDate=end_date,
                                            field=u"",pandas="1")[['isOpen','calendarDate']]
        calendar_date = calendar_date[calendar_date['isOpen']==1]
        calendar_date['calendarDate'] = pd.to_datetime(calendar_date['calendarDate'])
        trade_date = set(calendar_date['calendarDate'])
        if freq == 'monthly':
            calendar_date['year'] = calendar_date['calendarDate'].map(lambda x:x.year) 
            calendar_date['month'] = calendar_date['calendarDate'].map(lambda x:x.month)
            if point > 0:
                trade_date = set(calendar_date.groupby(['year','month']).head(point)['calendarDate']) - set(calendar_date.groupby(['year','month']).head(point-1)['calendarDate'])
            elif point < 0:
                point = - point
                trade_date = set(calendar_date.groupby(['year','month']).tail(point)['calendarDate']) - set(calendar_date.groupby(['year','month']).tail(point-1)['calendarDate'])
        elif freq == 'weekly':
            calendar_date['year_week'] = calendar_date['calendarDate'].map(lambda x:datetime.datetime.isocalendar(x)[0:2])
            if point > 0:
                trade_date = set(calendar_date.groupby(['year_week']).head(point)['calendarDate']) - set(calendar_date.groupby(['year_week']).head(point-1)['calendarDate']) 
            elif point < 0:
                point = - point
                trade_date = set(calendar_date.groupby(['year_week']).tail(point)['calendarDate']) - set(calendar_date.groupby(['year_week']).tail(point-1)['calendarDate'])
        elif freq == 'daily':
            trade_date = set(calendar_date['calendarDate'])
            return np.sort(list(trade_date))
        elif freq == 'quarterly':
            calendar_date['year'] = calendar_date['calendarDate'].map(lambda x:x.year) 
            calendar_date['month'] = calendar_date['calendarDate'].map(lambda x:x.month)
            calendar_date = calendar_date[calendar_date['month'].isin([4,8,10])]
            if point > 0:
                trade_date = set(calendar_date.groupby(['year','month']).head(point)['calendarDate']) - set(calendar_date.groupby(['year','month']).head(point-1)['calendarDate'])
            elif point < 0:
                point = - point
                trade_date = set(calendar_date.groupby(['year','month']).tail(point)['calendarDate']) - set(calendar_date.groupby(['year','month']).tail(point-1)['calendarDate'])
        return np.sort(list(trade_date))
    
    @staticmethod
    def step_trade_date(base_date, step, trading_days):
        """
        交易日向前向后漂移
        """
        base_date_index = trading_days.index(base_date)
        step_date_index = base_date_index + step
        return trading_days[step_date_index]

class SignalGenerate():
    def __init__(self,DateProc):
        self.factor_date = None
        self.price_date = None
        self.date_proc = DateProc
        self.trade_date = self.date_proc.trade_date
        self.benchmark_symbol = None
    
    def set_analysis_parameter(self,start_date,end_date,benchmark_symbol,freq=None,point=None):
        """
        这里的price_date是factor_date的下一交易日
        :param start_date: 起始日期
        :param end_date: 终止日期
        :param benchmark_symbol: 基准代码
        :param freq: 频率,日、周、月
        :param point: 取哪一天作为调仓日
        """
        self.factor_date = self.date_proc._get_date_list(start_date,end_date,freq,point)
        self.price_date = [self.date_proc.step_trade_date(date,1,self.trade_date) for date in self.factor_date]
        self.benchmark_symbol = benchmark_symbol
        
    def get_analysis_factor_df(self,factor_compute):
        """
        根据动态投资域，得到相应因子值
        :param factor_compute: 因子dataframe
        """
        factor_dict ={}
        for date in self.factor_date:
            universe = set_universe(self.benchmark_symbol,date=date)
            factor_dict[date] = factor_compute.get_factor(universe, date)
        factor_df = pd.DataFrame(factor_dict).T
        return factor_df
    
    def get_o2o_return(self,signal_df):
        """
        获取对应时间点的开盘价，计算开盘价买入、下一调仓日开盘价卖出之间的收益
        :param signal_df: 因子dataframe
        :return eq_return_df: 个股收益率dataframe
        """
        quotation_list = []
        universe = list(signal_df.columns)
        for date in self.price_date:
            universe = set_universe(symbol=self.benchmark_symbol,date=date)
            one_day_quotation = DataAPI.MktEqudAdjGet(tradeDate=date.strftime('%Y%m%d'),secID=universe,ticker=u"",isOpen="",
                                                      beginDate=u"",endDate=u"",field=u"secID,tradeDate,openPrice",pandas="1")
            # print(universe, date, one_day_quotation)
            quotation_list.append(one_day_quotation)
        quotation_df = pd.concat(quotation_list,axis=0)
        quotation_df['tradeDate'] = quotation_df['tradeDate'].map(lambda x:parse(x))
        quotation_df['secID'] = quotation_df['secID'].map(lambda x: str(x)[:6])
        quotation_df = quotation_df.set_index(['tradeDate','secID'])
        eq_return_df = quotation_df.unstack().replace(0,np.NaN).pct_change().shift(-1).iloc[0:-1].stack()
        eq_return_df = eq_return_df.rename(columns={'openPrice':'o2o_ret'})
        return eq_return_df
    
    def get_index_o2o_return(self):
        """
        获取指数开买、开卖间的收益率
        :return eq_return_df: 指数收益率dataframe
        """
        index_quotation_list = []
        for date in self.price_date:
            index_day_quotation = DataAPI.MktIdxdGet(tradeDate=date.strftime('%Y%m%d'),indexID=self.benchmark_symbol,ticker=u"",
                               beginDate=u"",endDate=u"",exchangeCD=u"XSHE,XSHG",field=u"tradeDate,openIndex",pandas="1")
            # print(index_day_quotation)
            index_quotation_list.append(index_day_quotation)
        index_quotation_df = pd.concat(index_quotation_list,axis=0)
        index_quotation_df['tradeDate'] = index_quotation_df['tradeDate'].map(lambda x:parse(x))
        return index_quotation_df.set_index(['tradeDate']).pct_change().shift(-1)
    
    def get_analysis_summary(self,factor_df,eq_return_df):
        """
        将因子df与收益df做合并
        TODO: 这里没考虑开盘涨停无法买入，开盘跌停无法卖出等case
        :param factor_df: 因子dataframe
        :param eq_return_df: 收益dataframe
        """
        factor_price_dict = {}
        for i,date in enumerate(self.factor_date):
            factor_price_dict[date] = self.price_date[i]
        summary_df = pd.concat([eq_return_df,factor_df.rename(index=factor_price_dict).stack()],axis=1).rename(columns={0:'factor_value'})
        summary_df = summary_df.dropna(subset=['factor_value'])
        return summary_df


class SignalAnalysis():
    def __init__(self,generate_class):
        self.signal_generate = generate_class
        self.index_return = self.signal_generate.get_index_o2o_return()

    def plot_signal_coverage(self,analysis_df,universe_lens,plot=True):
        """
        画出信号覆盖率
        :param analysis_df: 因子、收益dataframe
        :param universe_lens: 投资域内股票数量
        :param plot: 是否画图
        """
        signal_range = universe_lens
        siganl_coverage = analysis_df.groupby(analysis_df.index.get_level_values(0)).apply(lambda x:len(x.dropna()))
        if plot:
            fig = plt.figure(figsize=(18,7))
            ax = fig.add_subplot(111)
            ax.fill_between(siganl_coverage.dropna().index,0,siganl_coverage.dropna().values,color='skyblue')
            ax.set_ylim([0,signal_range])    
            ax.set_title(u"测试区间信号覆盖", fontproperties=font, fontsize=18)
            return np.mean(siganl_coverage) / universe_lens, siganl_coverage
        return np.mean(siganl_coverage) / universe_lens, siganl_coverage

    def plot_random_distribution_of_signal(self, factor_df):
        """
        随机选取9期因子数据，看分布，左偏、右偏等问题
        """
        fig, ax = plt.subplots(nrows=3, ncols=3, figsize=(20,15))
        for i in range(3):
            for j in range(3):
                rand = randint(0, len(factor_df) - 1)
                factor_df.iloc[rand].dropna().hist(bins=100,ax=ax[i][j])

    def factor_grouper(self,summary_df,quantile_point,col='factor_value'):
        """
        对因子进行分组
        :param summary_df: 因子、收益dataframe
        :param quantile_point: 分组
        :return quantile_df: 分组后的dataframe
        """
        def quantile_function(df, quantile_point, col='factor_value'):
            #Get the quantile value
            #Select the data in the group that falls at or below the quantile value and return it
            def group_num(x):
                if quantile_point >= 3:
                    if x <= df[col].quantile(1.0/quantile_point):
                        return 1
                    for i in range(2,quantile_point):
                        if x <=df[col].quantile(1.0*i/quantile_point) and x > df[col].quantile(1.0*(i-1)/quantile_point):
                            return i
                    if x > df[col].quantile(1.0*i/quantile_point):
                        return quantile_point
                elif quantile_point == 2:
                    if x <= df[col].quantile(1.0/quantile_point):
                        return 1
                    else:
                        return 2
                else:
                    raise ValueError('quantile group num must larger than 2!')
            df['group'] = df[col].map(lambda x:group_num(x))
            return df
        return summary_df.groupby(level=0).apply(quantile_function,quantile_point)
                
    def fama_macbeth_decay(self,factor_df,eq_return_df,decay=6):
        """
        因子对收益的预测能力
        :param factor_df: 因子dataframe
        :param eq_return_df: 个股收益dataframe
        :param decay: 滞后测试用到的期数
        :return decay_summary_df: decay分析结果
        """
        summary_df_list = []
        for i in range(decay):
            factor_df_ = factor_df.shift(i).dropna(how='all')
            summary_df = self.signal_generate.get_analysis_summary(factor_df_,eq_return_df)
            summary_df = summary_df.dropna(subset=['factor_value'])
            fm = pd.fama_macbeth(y=summary_df['o2o_ret'],x=summary_df[['factor_value']])
            summary_df_list.append([fm.mean_beta['factor_value'],fm.t_stat['factor_value'],i])
        return pd.DataFrame(summary_df_list).rename(columns={0:'mean_Coeff',1:'t_stat',2:'Decay'})
    
    def _compute_turnover(self,select_df):
        """
        计算换手率
        """
        select_df = select_df.unstack()
        accumulate_turnover = 0
        for i in range(len(select_df)-1):
            diff_universe = set(select_df.iloc[i].dropna().index) - set(select_df.iloc[i+1].dropna().index)
            dropna_len = len(select_df.iloc[i].dropna())
            accumulate_turnover = accumulate_turnover + 1.0*len(diff_universe)/dropna_len*2 if dropna_len != 0 else accumulate_turnover
        return accumulate_turnover / len(select_df) * 12
    
    def _compute_annual_return(self,ret_series):
        """
        计算年化收益
        """
        return gmean(ret_series.dropna()+1.)**12 - 1.
    
    def _compute_annual_std(self,ret_series):
        """
        计算年化收益标准差，也是风险的一种
        """
        return ret_series.dropna().std() * np.sqrt(12)
    
    def _compute_sharp(self,ret_series):
        """
        计算夏普率
        """
        return 1.0*self._compute_annual_return(ret_series) / self._compute_annual_std(ret_series)
    
    def _compute_mean_return(self,ret_series):
        """
        计算平均收益
        """
        return np.mean(ret_series)
    
    def _compute_sample_standard_deviation(self,ret_series):
        """
        计算收益的样本标准差
        """
        return np.std(ret_series, ddof=1)
    
    def long_only_ret(self,summary_df,benchmark,threhold=None,quantile_num=None):
        """
        仅做多收益
        :param summary_df: 因子、收益dataframe
        :param benchmark: 基准dataframe
        :param threhold: 百分比分组
        :param quantile_num: 分组数
        """
        def cut_big(x):
            if threhold is not None:
                x = x[x['factor_value']>x['factor_value'].quantile(threhold)]
                x.index = x.index.droplevel()
                return x
            elif quantile_num is not None:
                x = x[x['factor_value']>x['factor_value'].quantile(1 - 1.0/quantile_num)]
                x.index = x.index.droplevel()
                return x
            else:
                raise ValueError('you must set one of threhold or quantile_num!')
        summary_df_select = summary_df.groupby(summary_df.index.get_level_values(0)).apply(lambda x:cut_big(x))
        summary_df_mean = summary_df_select.groupby([summary_df_select.index.get_level_values(0)]).mean()
        if benchmark == 'equal':
            time_df_summary = summary_df_mean['o2o_ret'] - summary_df.groupby(summary_df.index.get_level_values(0))['o2o_ret'].mean()
        elif benchmark == 'index':
            time_df_summary = summary_df_mean['o2o_ret'] - self.index_return['openIndex']
        else:
            raise ValueError('benchmark only support equal or index !')
        return summary_df_select, time_df_summary
    
    def long_short_ret(self,summary_df,threhold=None,quantile_num=None):
        """
        多空收益
        :param summary_df: 因子、收益dataframe
        :param threhold: 百分比分组
        :param quantile_num: 分组数
        """
        def cut_df(x,flag):
            if threhold is not None:
                if flag == 'big':  
                    x = x[x['factor_value']>x['factor_value'].quantile(threhold, interpolation='nearest')]
                    x.index = x.index.droplevel()
                    return x
                elif flag == 'small':
                    x = x[x['factor_value']<x['factor_value'].quantile(threhold, interpolation='nearest')]
                    x.index = x.index.droplevel()
                    return x
            elif quantile_num is not None:
                if flag == 'big':
                    x = x[x['factor_value']>=x['factor_value'].quantile(1 - 1.0/quantile_num, interpolation='nearest')]
                    x.index = x.index.droplevel()
                    return x
                elif  flag == 'small':
                    x = x[x['factor_value']<=x['factor_value'].quantile(1.0/quantile_num, interpolation='nearest')]
                    x.index = x.index.droplevel()
                    return x
            else:
                raise ValueError('you must set one of threhold or quantile_num!')
        
        summary_df_big = summary_df.groupby(summary_df.index.get_level_values(0)).apply(lambda x:cut_df(x,'big'))
        summary_df_small = summary_df.groupby(summary_df.index.get_level_values(0)).apply(lambda x:cut_df(x,'small'))
        summary_df_big_mean = summary_df_big.groupby([summary_df_big.index.get_level_values(0)]).mean()
        summary_df_small_mean = summary_df_small.groupby([summary_df_small.index.get_level_values(0)]).mean()
        ret_summary = summary_df_big_mean['o2o_ret'] - summary_df_small_mean['o2o_ret']
        return summary_df_big, summary_df_small,ret_summary
    
    def get_decay_return_plot(self,factor_df,eq_return_df,ret_type,benchmark=None,threhold=None,quantile_num=None,decay=6):
        """
        收益的decay分析
        :param factor_df: 因子dataframe
        :param eq_return_df: 收益dataframe
        :param ret_type: long_only / long_short
        :param benchmark: 基准dataframe
        :param threhold: 百分比分组
        :param quantile_num: 分组数
        :param decay: 滞后期数
        """
        summary_df_list = []
        legend_col = []
        result_dict = {'年化收益':[],'标准差':[],'夏普率':[],'年化换手率':[]}   
        fig = plt.figure(figsize=(10,5))
        ax = fig.add_subplot(111)
        for i in range(decay):
            factor_df_ = factor_df.shift(i).dropna(how='all')
            summary_df = self.signal_generate.get_analysis_summary(factor_df_,eq_return_df)
            summary_df = summary_df.dropna(subset=['factor_value'])
            if ret_type == 'long_only':
                select_df, time_df_summary = self.long_only_ret(summary_df,quantile_num = quantile_num,threhold = threhold,benchmark=benchmark)
                result_dict['年化换手率'].append(self._compute_turnover(select_df['o2o_ret']))
                
            elif ret_type == 'long_short':
                select_big, select_small , time_df_summary = self.long_short_ret(summary_df,quantile_num = quantile_num,threhold = threhold)                      
                long_turnover = self._compute_turnover(select_big['o2o_ret'])
                short_turnover = self._compute_turnover(select_small['o2o_ret'])                            
                result_dict['年化换手率'].append(long_turnover + short_turnover)
            
            result_dict['年化收益'].append(self._compute_annual_return(time_df_summary))
            result_dict['标准差'].append(self._compute_annual_std(time_df_summary))
            result_dict['夏普率'].append(self._compute_sharp(time_df_summary))
            time_df_summary = (1+time_df_summary).cumprod()
            ax.plot(time_df_summary.index,time_df_summary.values,linewidth=2.5)
            legend_col.append('decay '+str(i))
            summary_df_list.append(time_df_summary)
        
        decay_summary_return = pd.concat(summary_df_list,axis=1)
        ax.legend(legend_col,loc='upper left',frameon= False)
        ax.spines['right'].set_color('none')
        ax.spines['top'].set_color('none')
        ax.xaxis.set_ticks_position('bottom')
        ax.yaxis.set_ticks_position('left')
        if ret_type == 'long_only':
            ax.set_title(u"测试区间long_only累计超额收益衰减", fontproperties=font, fontsize=18)
        elif ret_type == 'long_short':
            ax.set_title(u"测试区间long_short累计收益衰减", fontproperties=font, fontsize=18)
        result_df = pd.DataFrame(result_dict)
        result_df.index.name = 'decay'
        result_df = result_df[['年化收益','标准差','夏普率','年化换手率']]
        return decay_summary_return,result_df
    
    def get_long_short_return_stats(self,factor_df,eq_return_df,ret_type,benchmark=None,threhold=None,quantile_num=None):
        """
        计算平均收益、标准差、t统计量(关于什么的t统计)get_long_short_return_stats
        :param factor_df: 因子dataframe
        :param eq_return_df: 收益dataframe
        :param ret_type: long_only / long_short
        :param benchmark: 基准dataframe
        :param threhold: 百分比分组
        :param quantile_num: 分组数
        """
        result_dict = {'平均对冲收益':[],'对冲收益标准差':[]} 
        factor_df_ = factor_df.dropna(how='all')
        summary_df = self.signal_generate.get_analysis_summary(factor_df_,eq_return_df)
        summary_df = summary_df.dropna(subset=['factor_value'])
        if ret_type == 'long_only':
            select_df, time_df_summary = self.long_only_ret(summary_df,quantile_num=quantile_num,threhold=threhold,benchmark=benchmark)
        elif ret_type == 'long_short':
            select_big, select_small , time_df_summary = self.long_short_ret(summary_df,quantile_num=quantile_num,threhold=threhold)
        result_dict['平均对冲收益'].append(self._compute_mean_return(time_df_summary))
        result_dict['对冲收益标准差'].append(self._compute_sample_standard_deviation(time_df_summary))
        result_df = pd.DataFrame(result_dict)
        result_df.index.name = 'decay'
        result_df = result_df[['平均对冲收益','对冲收益标准差']]
        return result_df     
    
    def get_quantile_cumulative_return_plot(self,summary_df,quantile_num,benchmark):
        """
        分组累计超额收益
        :param summary_df: 因子、收益dataframe
        :param quantile_num: 分组数
        :param benchmark: 基准dataframe
        """
        # if out_benchmark:
        #     index_o2o_return = self.get_index_o2o_return()
        group_summary_df = self.factor_grouper(summary_df,quantile_num)
        time_group_df = group_summary_df.groupby([group_summary_df.index.get_level_values(0),'group']).mean()
        # print time_group_df
        def get_return(x):
            x.index = x.index.get_level_values(0)
            if benchmark == 'equal':
                return x['o2o_ret'] - summary_df.groupby(summary_df.index.get_level_values(0))['o2o_ret'].mean()
            elif benchmark == 'index':
                return x['o2o_ret'] - self.index_return['openIndex']
            else:
                raise ValueError('benchmark only support equal or index !')
        time_group_df_summary = time_group_df.groupby(time_group_df.index.get_level_values(1)).apply(lambda x:get_return(x)).T
        ret = time_group_df_summary.apply(lambda x:self._compute_annual_return(x))
        sigma = time_group_df_summary.apply(lambda x:self._compute_annual_std(x))
        result = pd.DataFrame([ret,sigma],index = ['年化收益','标准差']).T
        result['夏普率'] = result['年化收益'] / result['标准差']
        result['年化换手率'] = group_summary_df.groupby(['group']).apply(lambda x:self._compute_turnover(x['o2o_ret']))
        result.index.name = 'group'
        time_group_df_summary_ = (1 + time_group_df_summary).cumprod()
        fig = plt.figure(figsize=(10,5))
        ax = fig.add_subplot(111)
        for group in time_group_df_summary_.columns:
            ax.plot(time_group_df_summary_.index,time_group_df_summary_[group].values,linewidth=2.5)
        ax.legend(time_group_df_summary_.columns,loc='upper left',frameon= False)
        ax.spines['right'].set_color('none')
        ax.spines['top'].set_color('none')
        ax.xaxis.set_ticks_position('bottom')
        ax.yaxis.set_ticks_position('left')
        ax.set_title(u"测试区间分位数分组累计超额收益", fontproperties=font, fontsize=18)
        return time_group_df_summary_ , result
    
    def get_quantile_return_mean_plot(self,summary_df,quantile_num,plot=True):
        """
        分组平均超额收益
        :param summary_df: 因子、收益dataframe
        :param quantile_num: 分组数
        :param plot: 是否画图
        """
        group_summary_df = self.factor_grouper(summary_df,quantile_num)
        time_group_df = group_summary_df.groupby([group_summary_df.index.get_level_values(0),'group']).mean()
        time_group_df_summary = time_group_df.groupby(time_group_df.index.get_level_values(1)).mean()
        if plot:
            fig = plt.figure(figsize=(8,4))
            ax = fig.add_subplot(111)
            time_group_df_summary['o2o_ret'].plot(kind='bar',ax=ax,color='deepskyblue')
            ax.spines['top'].set_color('none')
            ax.xaxis.set_ticks_position('bottom')
            ax.spines['right'].set_color('none')
            ax.yaxis.set_ticks_position('left')
            ax.spines['bottom'].set_position('zero')
            ax.set_title(u"测试区间分组平均收益", fontproperties=font, fontsize=18)
            ax.grid()
        return time_group_df_summary

    def get_ic_series(self,fator_return_df):
        """
        返回IC序列
        :param fator_return_df: 因子、收益dataframe
        """
        def omit_nan_spearmanr(x):
            x = x.dropna()
            if len(x) > 0:
                return st.spearmanr(x['o2o_ret'],x['factor_value'])
        ic_result = fator_return_df.groupby(level=0).apply(lambda x:omit_nan_spearmanr(x))
        ic_result = pd.DataFrame(ic_result.dropna())
        ic_result['IC'] = ic_result[0].map(lambda x:x[0])
        ic_result['IC_Pvalue'] = ic_result[0].map(lambda x:x[1])
        return ic_result[['IC','IC_Pvalue']]
    
    def get_ic_plot(self,fator_return_df):
        """
        ic分析并画图
        :param fator_return_df: 因子、收益dataframe
        """
        ic_series = self.get_ic_series(fator_return_df)
        fig, ax = plt.subplots(nrows=1, ncols=2,figsize=(20,5))
        ax[0].plot(ic_series.index,ic_series['IC'],'o-',color='cyan')
        ax[0].plot(ic_series.index,pd.rolling_mean(ic_series['IC'],10),color = 'deepskyblue',linewidth = 2.5)
        ax[1].plot(ic_series.index,ic_series['IC_Pvalue'],'v-',color = 'lightgreen')
        ax[1].plot(ic_series.index,[0.05]*len(ic_series),'r',linewidth=2)
        ax[0].legend(['IC','MA10_IC'],loc='upper left',prop=font,frameon= False,handlelength = 3,handletextpad=0.1,borderpad = 0.1)
        ax[1].legend(['IC_Pvalue','Pvalue_0.05'],loc='upper left',prop=font,frameon= False,handlelength = 3,handletextpad=0.1,borderpad = 0.1)
        def myfunc(x, pos=0): 
            return "{0:.2f}%".format(x * 100)
        ax[1].yaxis.set_major_formatter(FuncFormatter(myfunc))
        ax[0].set_ylabel('IC')
        ax[1].set_ylabel('IC_Pvalue')
        ax[0].set_title(u"IC时序图", fontproperties=font, fontsize=18)
        ax[1].set_title(u"IC_Pvalue时序图", fontproperties=font, fontsize=18)
        ax[0].grid()
        ax[1].grid()
        return ic_series
```

将前两步计算所得的单因子，进行合成。其中，会考虑因子的方向、去极值、归一化、正态化等。

该部分内容为：

- 对合成后的大类因子进行归一化、等权合成

- 测试合成后的因子，与市值、质量、动量风格、ROE、股息率因子的相关性

```python
from scipy.stats import spearmanr
import matplotlib.pyplot as plt
# CAL.PyCAL中包含font
from CAL.PyCAL import *
from itertools import cycle

def winsorize_in_paper(factor, direction=1.0):
    """
    去极值
    """
    factor_new = factor.copy(deep=True)
    factor_new = factor_new * direction
    for _, rows in factor_new.iterrows():
        f_std = np.std(rows)
        f_mean = np.mean(rows)
        above_col = rows[rows > (f_mean + 3 * f_std)].index
        below_col = rows[rows < (f_mean - 3 * f_std)].index
        factor_new.loc[_, above_col] = f_mean + 3 * f_std
        factor_new.loc[_, below_col] = f_mean - 3 * f_std
    return factor_new

def max_min_normalize(factor):
    """
    归一化
    """
    factor_new = factor.copy(deep=True)
    for _, rows in factor_new.iterrows():
        f_max = np.max(rows)
        f_min = np.min(rows)
        if f_max - f_min > 0.0:
            factor_new.loc[_] = (rows - f_min) / (f_max - f_min)
    return factor_new

def normalize_in_paper(factor):
    """
    正态化
    """
    factor_new = factor.copy(deep=True)
    for _, rows in factor_new.iterrows():
        f_std = np.std(rows)
        f_mean = np.mean(rows)
        if f_std > 0.0:
            factor_new.loc[_] = (rows - f_mean) / f_std
    return factor_new

def check_factor_correlation(factor):
    """
    检查合成后因子与市值、质量、动量风格、ROE、股息率因子的相关性
    """
    data = []
    test_factors = ['ROE', 'NLSIZE', 'REVS60', 'DividendPS']
    for _, rows in factor.iterrows():
        td = _.strftime('%Y%m%d')
        factor_df = rows.to_frame().reset_index()
        factor_df = factor_df.T.reset_index(drop=True).T
        factor_df.columns = ['ticker', 'ESG']
        tickers = rows.index.values.tolist()
        other_factor_df = DataAPI.MktStockFactorsOneDayProGet(tradeDate=td,secID=u"",ticker=",".join(tickers),field=','.join(['ticker', 'tradeDate'] + test_factors),pandas="1")
        other_factor_df = other_factor_df.merge(factor_df, left_on=['ticker'], right_on=['ticker'], how='inner', suffixes=['', '_'])
        # print(other_factor_df.head(5))
        for other_factor_name in test_factors:
            valid_df = other_factor_df[[other_factor_name, 'ESG']].dropna(how='any', axis=0)
            data.append([td, other_factor_name, spearmanr(valid_df[other_factor_name], valid_df['ESG'])[0]])
        # print('[CorrTest]td: {0} corr computed done'.format(td))
        # break
    corr_df = pd.DataFrame(data, columns=['tradeDate', 'otherFactorName', 'corr'])
    # 画图
    cycol = cycle('bgrcmk')
    fig = plt.figure(figsize=(10,5))
    ax = fig.add_subplot(111)
    for factor_name in test_factors:
        part_df = corr_df[corr_df['otherFactorName'] == factor_name].copy(deep=True).set_index('tradeDate')
        part_df['corr'].plot(label=factor_name, ax=ax, alpha=0.5, color=cycol.next())
    ax.legend()
    ax.set_title(u"ESG与相关因子的相关性", fontproperties=font, fontsize=18)
    return corr_df

factors_type = {'board': {'BOARD_SIZE': -1.0, 'IND_DIR_AVG_EDU': 1.0, 'IND_DIR_PART_RATIO': -1.0, 'SUPERVISOR_NUM': 1.0,
                          'COMMITTEE_NUM': 1.0, 'CEO_CHANGE_NUM': -1.0, 'CHAIRMAN_CHANGE_NUM': -1.0, 'IND_DIR_NUM_RATIO': 1.0,
                          'IND_DIR_FIN_RATIO': 1.0, 'IND_DIR_LAW_RATIO': 1.0, 'IND_DIR_ATTEND_RATIO': 1.0, 'IND_DIR_ABOVE_SIXTY': -1.0},
                'ownership': {'TOP_SHARE_HOLDER_PCT': -1.0, 'ACTUAL_CONTROLLER_HOLDER_PCT': -1.0, 'CONTROL_HOLDER_PCT': -1.0, 'SUPERVISOR_HOLDER_PCT': 1.0,
                             'MANAGER_HOLDER_PCT': 1.0, 'NORMAL_MANAGER_HOLDER_PCT': 1.0, 'FUND_HOLD_RATIO_PCT': 1.0, 'SSF_SHARE_HOLDER_PCT': 1.0,
                             'QFII_SHARE_HOLDER_PCT': 1.0, 'WITH_HK_SHARE': 1.0},
                'finance': {'AFS': 1.0, 'AF_CHANGE': -1.0, 'BIG_FOUR': 1.0, 'INTERNAL_CONTROL': 1.0, 'EE_PROFIT_ERROR': -1.0, 'ST_FLAG': -1.0},
                'pay': {'DIR_ANN_REWARD': 1.0, 'MANAGER_ANN_REWARD': 1.0, 'SHARE_EXCIT_RATIO': 1.0},
                'behavior': {'DIR_SUP_PLEDGE_RATIO': -1.0, 'ADMIN_EXP_RATIO': -1.0, 'ASSET_TURNOVER': 1.0, 'OTHER_RECEIVE_RATIO': -1.0, 'RELATED_SALES_RATIO': -1.0,
                             'GUARANTEE_RATIO': -1.0, 'MERGER_ACQUISITION_NUM': -1.0, 'MERGER_ACQUISITION_SPLIT_NUM': -1.0, 'ALLOT_SHARES_RATIO': -1.0,
                             'GOODWILL_RATIO': -1.0, 'MERGER_ACQUISITION_FAIL_NUM': -1.0, 'COMP_ILLEGAL_COUNT': -1.0, 'SUIT_AMOUNT_RATIO': -1.0,
                             'FREEZE_RATIO': -1.0, 'INVEST_ACTIVITY_NUM': 1.0, 'INVEST_ACTIVITY_ORG_NUM': 1.0}}

# factors_type = {'board': {'BOARD_SIZE': -1.0}}

# 进行单因子测试
flag, multi_factor_df = load_pickle('{0}/esg_multi_factor_test.data'.format(dir_path), ori_obj=pd.DataFrame())
if not flag:
    t1 = time.time()
    # 1) 大类因子内做预处理
    cluster_factor_dic = {}
    multi_factor_df = pd.DataFrame()
    for idx, f_type in enumerate(factors_type):
        cluster_factor_dic[f_type] = pd.DataFrame()
        for f_name in factors_type[f_type]:
            if f_name not in factors:
                continue
            factor_df = factors[f_name]
            factor_df = winsorize_in_paper(factor_df, direction=factors_type[f_type][f_name])
            factor_df = max_min_normalize(factor_df)
            cluster_factor_dic[f_type] = cluster_factor_dic[f_type] + factor_df if len(cluster_factor_dic[f_type].index) > 0 else factor_df
        # 2) 大类因子预处理
        cluster_factor_dic[f_type] = winsorize_in_paper(cluster_factor_dic[f_type], 1.0)
        cluster_factor_dic[f_type] = normalize_in_paper(cluster_factor_dic[f_type])
        multi_factor_df = multi_factor_df + cluster_factor_dic[f_type] if len(multi_factor_df) > 0 else cluster_factor_dic[f_type]
    save_pickle('{0}/esg_multi_factor_test.data'.format(dir_path), result_df)
    print('[Multi Factor] mult factor merged done. cost: {0:.2f}'.format(time.time()-t1))

# print(multi_factor_df.head(5))
# 3) 检验合成因子与常见风格因子间的相关性
corr_df = check_factor_correlation(multi_factor_df)
```
