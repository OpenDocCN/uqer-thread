# 三层因子检测模型-中期因子与短期因子

```python
import pandas as pd
import numpy as np
import datetime
import time
from scipy.stats import spearmanr
```

```python
today = time.strftime("%Y-%m-%d")
```

```python
## 获取交易日历
def get_celedate(begin = '2010-01-01', end = today, symbol = 'day'):
    df_date = DataAPI.TradeCalGet(exchangeCD=u"XSHG,XSHE", field=u"")
    df_date = df_date[(df_date['calendarDate'] > begin) & (df_date['calendarDate'] < end)]
    df_date = df_date[df_date['isOpen'] == 1]
    if symbol == 'day':
        return df_date['calendarDate']
    if symbol == 'month':
        return df_date[df_date['isMonthEnd'] == 1]['calendarDate']
    if symbol == 'week':
        return df_date[df_date['isWeekEnd'] != 1]['calendarDate']
    if symbol == 'all':
        return df_date
```

```python
## 获取行业分类
def get_ind_class():
    df_ind = DataAPI.EquIndustryGet(secID=u"",ticker=u"",industryVersionCD=u"010303",industry=u"",industryID1=u"",industryID2=u"",industryID3=u"",intoDate=u"",field=u"",pandas="1")
    dict_ind = {k:v for k, v in df_ind[df_ind['isNew'] == 1][['ticker', 'industryName1']].values}
    return dict_ind
dict_ind = get_ind_class()
```

```python
## 计算因子之间spearman相关系数
def get_corr(ser1, ser2):
    val, p = spearmanr(ser1, ser2)
    return p, val
```

```python
date_pre = '2018-01-03'
date_late = '2018-02-01'
```

```python
def get_fac_test(date_pre, date_late):
    ## 获取数据
    df_fac = DataAPI.MktStockFactorsOneDayGet(tradeDate= date_pre ,secID=u"",ticker=u"",field=u"",pandas="1")
    df_price_pre = DataAPI.MktEqudGet(secID=u"",ticker=u"",tradeDate= date_pre ,beginDate=u"",endDate=u"",isOpen="",field=u"ticker,closePrice",pandas="1")
    df_price_late = DataAPI.MktEqudGet(secID=u"",ticker=u"",tradeDate= date_late ,beginDate=u"",endDate=u"",isOpen="",field=u"ticker,closePrice",pandas="1")
    
    df_price = pd.merge(df_price_pre, df_price_late, on= 'ticker', how= 'inner')
    df_price['ratio'] = (df_price['closePrice_y'] - df_price['closePrice_x'])  / df_price['closePrice_x']
    dict_ratio = {k:v for k,v in df_price[['ticker', 'ratio']].values}
    ## 填充收益与行业分类
    df_fac['ratio'] = df_fac['ticker'].map(dict_ratio)
    df_fac['ind'] = df_fac['ticker'].map(dict_ind)
    df_group = df_fac.groupby('ind').apply(lambda x: x[:])
    ## 计算因子与行业之间的相关系数，进行检验
    fit_col = [u'secID', u'ticker', u'tradeDate', u'ratio', u'ind']
    fac_list = [i for i in df_fac.columns if i not in fit_col] 
    new_df = pd.DataFrame()
    for i in df_group.index.levels[0]:
        df_i = df_group.loc[i]
        for j in fac_list:
            x, y = df_i[[j, 'ratio']].dropna().values.T
            val, p = spearmanr(x, y)
            if p > 0.2:
                val = 0
            new_df.loc[i, j] = val
    new_df['date'] = date_pre
    new_df['ind'] = new_df.index
    new_df.index = range(new_df.shape[0])
    return new_df
```

```python
df_test = get_fac_test(date_pre, date_late)
```

```python
fit_col = [u'secID', u'ticker', u'tradeDate', u'ratio', u'ind']
fac_list = [i for i in df_test.columns if i not in fit_col] 
```

```python
def fac_testdays(num_day, symbol = 'month'):
    df_test_day = pd.concat([get_fac_test(date_pre, date_late) for date_pre, date_late in zip(get_celedate(symbol=symbol)[-num_day-1:-1], get_celedate(symbol= symbol )[-num_day:])])
    return df_test_day
df_test_day = fac_testdays(20)
```

```python
def test_effect_fac(ser):
    c = ser.shape[0]
    zero_c = ser[ser == 0].shape[0]
    non_c = ser[ser < 0].shape[0]
    pol_c = ser[ser > 0].shape[0]
    if zero_c / c < 0.5:
        return False
    # else:
    #     if np.abs(pol_c - non_c)/ (pol_c + non_c) < 0.2:
    #         return False
    else:
        return True
```

```python
def get_ind_effect_fac(df_test_day):
    df_g_test = df_test_day.groupby(['ind','date']).apply(lambda x: x[:])[fac_list]
    dict_effect_cols = {}
    for name in df_g_test.index.levels[0]:
        cols = df_g_test.columns[df_g_test.loc[name].apply(lambda x:test_effect_fac(x), axis = 0).values]
        dict_effect_cols[name] = list(cols)
    return pd.Series(dict_effect_cols)
```

```python
## 检测中期因子
df_test_day = fac_testdays(20, symbol= 'month')
get_ind_effect_fac(df_test_day)
```

```python
## 检测短期因子
df_test_day = fac_testdays(20, symbol= 'week')
get_ind_effect_fac(df_test_day)
```
