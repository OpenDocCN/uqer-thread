# **一、奇异谱分析**

## **1.1 什么是奇异谱分析**

奇异谱分析是近年来兴起的一种研究非线性时间序列数据的强大的方法。它根据所观测到的时间序列构造出轨迹矩阵，并对轨迹矩阵进行分解、重构，从而提取出代表原时间序列不同成分的信号，如长期趋势信号、周期信号、噪声信号等，从而对时间序列的结构进行分析，并可进一步预测。

奇异谱分析(SSA)方法最早由colebrook于1978年首先在海洋学研究中提出并使用。Fracrich用一维时间序列在延迟相空间中做EOF展开，再通过显著性检验研究确定有意义的特征成分的个数，据此估计气候吸引子的维数。这个工作被认为是SSA在气象学中的最早应用。Hassani将这种方法引人到社会问题研究中来，并用其预测了美国交通事故的月时间序列数据。N.Golyandina给出了奇异谱分析的扩展形式一多通道奇异谱分析的算法，并由Hossein Hassani用来对英镑/美元汇率进行了分析预测，取得了较好的效果。

奇异谱分析的基本思想是，将所观测到的一维时间序列数据$$Y(T) = (y(1),\cdots ,y(T))$$ 转化为其轨迹矩阵$$X = (x_{ij})_{i,j = 1}^{n,m} = \begin{pmatrix} y_1 & y_2 & y_3 & \cdots & y_m \\ y_2 & y_3 & y_4 & \cdots & y_{m+1} \\ \vdots & \vdots & \vdots & \cdots & \vdots \\ y_n & y_{n+1} & y_{n+2} & \cdots & y_T \end{pmatrix}$$
其中，m为选取的窗口长度，n=T-m+1，计算X.T\*X并对其进行奇异值分解（SVD），从而得到其m个特征值$$\lambda_1\geq\lambda_2\geq\cdots\geq\lambda_m\geq0 $$及其相应的特征向量将每一个特征值所代表的信号进行分析组合，重构出新的时间序列。奇异谱分析过程可分成嵌人、SVD分解、分组、重构四个步骤，下面详细介绍具体算法。 

## **1.2 奇异谱分析的算法**

### **1.嵌入**

选择适当的窗口长度$$m(2 \leq m \leq T)$$ 将所观测到的一维金融时间序列数据转化为多维序列$$X_1 , \cdots , X_n ,(X_i = (y_i,\cdots,y_{i+m-1})  , n = T-m+1)$$ 得到轨迹矩阵$$X = [X_1 , \cdots , X_n] = (x_ij)_{i,j=1}^{n,m}$$ 这里m的选取不宜超过整个数据长度的1/3，如可根据事先经验大致确定数据的周期特征，则m的选取最好为周期的整数倍。<br>

## **1.3 对角平均法重构原序列**

假设经过svd分解重构后的矩阵为:$$X = (x_{ij})_{i,j = 1}^{n,m} = \begin{pmatrix} x_{1,1} & x_{1,2} &x_{1,3} & \cdots & x_{1,m} \\ x_{2,1} & x_{2,2} & x_{2,3} & \cdots & x_{2,m+1} \\ \vdots & \vdots & \vdots & \cdots & \vdots \\ x_{n,1} & x_{n,2} & x_{n,3} & \cdots & x_{n,n+m-1} \end{pmatrix}$$<br>
则重构序列，$$ y_1 = x_{1,1}  $$ $$ y_2 = (x_{1,2}+x_{2,1})/2 $$  $$ y_3 =(x_{1,3}+x_{2,2}+x_{3,1})/3 $$ $$ \cdots $$ $$y_T = x_{n,n-m+1}$$

写成公式:
$$ y_p = (∑_{j=1}^{m}X^{i,j})/α_p$$ $$ i = p-j+1$$ $$ 0 < i < n+1$$ $$ α_p =\begin{cases}p& p < m ,\\T-p+1& p > T-m+1,\\m& otherwise\end{cases}$$

## **1.4代码实现SSA的计算**
代码中按造研报内容，取第一奇异向量

```python
from __future__ import division
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model

# 嵌入
def getWindowMatrix(inputArray,t,m):
    temp = []
    n = t - m + 1
    for i in range(n):
        temp.append(inputArray[i:i+m])
    WindowMatrix = np.array(temp)
    return WindowMatrix

# 奇异谱分析，取第一主成分分量，返回重构矩阵
def SVDreduce(WindowMatrix):
    u,s,v = np.linalg.svd(WindowMatrix) #svd分解
    m1,n1 = u.shape
    m2,n2 = v.shape
    index = s.argmax()
    u1 = u[:,index]
    v1 = v[index]
    u1 = u1.reshape((m1,1))
    v1 = v1.reshape((1,n2))
    value = s.max()
    newMatrix = value*(np.dot(u1,v1))  #重构矩阵
    return newMatrix

# 对角线平均法重构序列
def recreateArray(newMatrix,t,m):
    ret = []
    n = t - m +1
    for p in range(1,t+1):
        if p<m:
            alpha = p
        elif p>t-m+1:
            alpha = t-p+1
        else:
            alpha = m
        sigma = 0
        for j in range(1,m+1):
            i = p - j +1
            if i>0 and i<n+1:
                sigma += newMatrix[i-1][j-1]
        ret.append(sigma/alpha)
    return ret

# 按不同的序列、不同的窗口大小计算SSA
def SSA(inputArray,t,m):
    WindowMatrix = getWindowMatrix(inputArray,t,m)
    newMatrix    = SVDreduce(WindowMatrix)
    newArray     = recreateArray(newMatrix,t,m)
    return newArray

# MA均线计算
def MA(inputArray,m):
    output = []
    length = len(inputArray)
    for i in range(m):
        output.append(inputArray[0])
    for i in range(m,length):
        temp = inputArray[i-m:i]
        output.append(temp.sum()/m)
    return output
```

# **二、 SSA均线和MA均线的比较**

## **2.1 对末段点刻画的比较**

以窗口参数为3的SSA为例，假设有价格序列[x1,x2,x3,x4,x5]，对应移动窗口矩阵如下：
$$H = \begin{bmatrix} x_1 & x_2 & x_3 \\ x_2 & x_3 & x_4 \\ x_3 & x_4 & x_5 \end{bmatrix}\quad$$
假设对其协方差阵奇异值分解得到的第一主成分方向为:
$$\begin{bmatrix} a_1 & a_2 & a_3 \end{bmatrix}^{T}\quad$$
则重构矩阵为:
$$\widehat H = \begin{bmatrix} x_1 & x_2 & x_3 \\ x_2 & x_3 & x_4 \\ x_3 & x_4 & x_5 \end{bmatrix} \begin{bmatrix} a_1 & a_2 & a_3 \end{bmatrix} \begin{bmatrix} a_1 \\ a_2 \\ a_3 \end{bmatrix}  $$
$$ = \begin{bmatrix} a_1^{2}x_1+a_1a_2x_2+a_1a_3x_3 & a_1a_2x_1+a_2^{2}x_2+a_2a_3x_3 & a_1a_3x_1+a_2a_3x_2+a_3^{2}x_3 \\ a_1^{2}x_2+a_1a_2x_3+a_1a_3x_4& a_1a_2x_2+a_2^{2}x_3+a_2a_3x_4 & a_1a_3x_2+a_2a_3x_3+a_3^{2}x_4  \\ a_1^{2}x_3+a_1a_2x_4+a_1a_3x_5 & a_1a_2x_3+a_2^{2}x_4+a_2a_3x_5 & a_1a_3x_3+a_2a_3x_4+a_3^{2}x_5  \end{bmatrix}\quad$$
从而对角平均求出趋势线为：
$$\widehat x_1 = a_1^{2}x_1+a_1a_2x_2+a_1a_3x_3$$
$$\widehat x_2 = [a_1a_2x_1+(a_1^{2}+a_2^{2})x_2+(a_1a_2+a_2a_3)x_3+a_1a_3x_4]/2 $$
$$\widehat x_3 = [a_1a_3x_1+(a_1a_2+a_2a_3)x_2+(a_1^{2}+a_2^{2}+a_3^{2})x_3+(a_1a_2+a_2a_3)x_4+a_1a_3x_5]/3 $$
$$\widehat x_4 = [a_1a_3x_2+(a_1a_2+a_2a_3)x_3+(a_2^{2}+a_3^{2})x_4+a_2a_3x_5]/2 $$
$$\widehat x_5 = a_1a_3x_3+a_2a_3x_4+a_3^{2}x_5$$

从择时策略的角度来说，最关心的是**趋势线末端的形态和变化**，要及时准确的抓住市场的趋势离不开对趋势线末端的精准判断。SSA趋势线的末端与MA趋势线的末端不同。从以窗口参数为3的MA和SSA示例中可看到MA末端点由价格序列末3项等权平均组成，SSA末端点同样由价格序列末3项构成，但权重大小**和参数 a1,a2,a3 的大小对应，具有时变特征**。

奇异谱分析趋势线的特点在于，在上涨趋势中，第一主成分方向参数大小关系为a1 < a2 < a3 ，下跌趋势中，参数大小关系为 a1 > a2 > a3，从而对应的趋势线末端点$$\widehat x_5 = a_1a_3x_3+a_2a_3x_4+a_3^{2}x_5$$各个构成项权重大小表现出在**上涨趋势中新数据比旧数据权重大（权重递增）**，**下跌趋势中旧数据比新数据权重大（权重递减）的特点**。简而言之，由于MA对窗口内参数是等权重的，权重时变的SSA末端点在**上涨趋势中对回调的敏感性相对MA更强，下跌趋势中对反弹的确认相对MA更谨慎**。

在研报<基于奇异谱分析的均线择时研究>中，对SSA序列和MA序列作了模拟测试，在叠加服从高斯分布的白噪声序列后，模拟10000次，对于先增后减序列，MA5末端有2411例呈下跌趋势，SSA5末端有7204例呈下跌趋势，对于先减后增序列，MA5末端有7583例呈上涨趋势，SSA5端有1734例呈上涨趋势。表明在噪声影响下，SSA相比MA仍能反映出对**回调的高敏感性**和对**反弹的谨慎性**。下面我们简单测试下效果：

```python
# 回测,自己手写，比较粗糙，轻拍
m = 5
t = 2*m
netValue_ssa = 1 #净值 初始化1
netValue_ma = 1
closeIndex = np.array(data['closeIndex']) # HS300指数收盘
openIndex = np.array(data['openIndex'])
datelist  = np.array(data['tradeDate']) #日期

flag_ssa = False # 持有标志，买入持仓则为True,否则为False
flag_ma = False

strategy_ssa = {}
strategy_ma ={}
length = len(closeIndex)
startHold_ssa =[] #记录多头当期成本价及当期净值
startHold_ma =[]

# 假设每天开盘后以昨天的收盘价判断；但以今天的开盘价成交
for i in range(t,length):# 前t天不动作
    yesterday = datelist[i-1]
    today = datelist[i]
    inputArray = closeIndex[:i]
    inputArray = inputArray[-t:]
    SSA5 = SSA(inputArray,t,m) # m=5
    MA5  = MA(inputArray,m)
    valueYesterday = closeIndex[i-1] #看上个交易日的收盘价
    valueToday = openIndex[i] # 若买入，为买入价
    
    ############################################   ssa   ############################################
    # 计算当前净值(截止昨天收盘)ssa，因此最后一天的净值不算进去，实际回测天数相当于少一天，whatever~
    if len(startHold_ssa)!=0: #持有头寸
        netValue_ssa = (startHold_ssa[0])*(valueYesterday/startHold_ssa[1])  # 这里是计算当前值与买入时成本的变化
    else: #空仓时
        netValue_ssa = netValue_ssa
    strategy_ssa.update({yesterday:netValue_ssa})
    
    # strategy_ssa判断买入 
    if flag_ssa==False and valueYesterday>SSA5[-1] and closeIndex[i-2]<SSA5[-2]:
        flag_ssa = True
        startHold_ssa = [netValue_ssa,valueToday]
    
    # strategy_ssa判断卖出
    if flag_ssa==True and valueYesterday<SSA5[-1] and closeIndex[i-2]>SSA5[-2]:
        netValue_ssa = (startHold_ssa[0])*(valueToday/startHold_ssa[1])  #以开盘价作为卖出价计算当前净值
        flag_ssa = False
        startHold_ssa = [] # 清仓
    #################################################################################################

    #$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$       ma       $$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$
    # 计算当前净值(截止昨天收盘)ma
    if len(startHold_ma)!=0: #持有头寸
        netValue_ma = (startHold_ma[0])*(valueYesterday/startHold_ma[1])  # 这里是计算当前值与买入时成本的变化
    else: #空仓时
        netValue_ma = netValue_ma
    strategy_ma.update({yesterday:netValue_ma})
    
    # strategy_ma判断买入 
    if flag_ma==False and valueYesterday>MA5[-1] and closeIndex[i-2]<MA5[-2]:
        flag_ma = True
        startHold_ma = [netValue_ma,valueToday]
    
    # strategy_ma判断卖出
    if flag_ma==True and valueYesterday<MA5[-1] and closeIndex[i-2]>MA5[-2]:
        netValue_ma = (startHold_ma[0])*(valueToday/startHold_ma[1])  #以开盘价作为卖出价计算当前净值
        flag_ma = False
        startHold_ma = [] # 清仓
    #$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$
```

从净值曲线可以看出，SSA单均线策略是完爆MA均线的，具体的，我们可以看看策略的表现：

```python
# 指标计算
# 年华收益
def AnnualizedReturns(pStart,pEnd,n):  #pStart为初始净值，pEnd为最终净值，n为策略持续时间（交易日）
    return (pEnd/pStart)**(250/n) -1

# 最大回撤（输入为list或array）
def MaxDrawdown(strategy):
    length = len(strategy)
    drawdown = []
    py =strategy[0]
    for i in range(1,length):
        px = strategy[i]
        py = max(strategy[:i])
        drawdown.append(1-(px/py))
    return max(drawdown)
                
```

```python
MD_ssa5 = MaxDrawdown(np.array(strategy_cmp['strategy_ssa5']))# 最大回撤
MD_ma5 = MaxDrawdown(np.array(strategy_cmp['strategy_ma5']))
AR_ssa5 = AnnualizedReturns(1,strategy_cmp['strategy_ssa5'][-2:-1].values[0],len(strategy_cmp))# 年化收益
AR_ma5 = AnnualizedReturns(1,strategy_cmp['strategy_ma5'][-2:-1].values[0],len(strategy_cmp))
strategy_performance = pd.DataFrame(columns=['最终净值','年化收益','最大回撤'],index=['strategy_ssa5','strategy_ma5'])
strategy_performance['最终净值']=[strategy_cmp['strategy_ssa5'][-2:-1].values[0],strategy_cmp['strategy_ma5'][-2:-1].values[0]]
strategy_performance['年化收益']=[AR_ssa5,AR_ma5]
strategy_performance['最大回撤']=[MD_ssa5,MD_ma5]
strategy_performance
```

MA单均线策略最终净值132.57%，年化收益率3.17%，最大回撤-52.05%，SSA单均线策略最终净值211.70%，年化收益率8.66%，最大回撤-39.09%。在2011年到2014年上半年的震荡市中，明显可以看到SSA替代MA提高了策略效果。**由于SSA在震荡市中能够更快速的确认趋势结束，因此由小优势慢慢累积可形成较大的相对优势。但是对趋势结束的高敏感性同样会导致策略在市场有大趋势的情况下会因为一些趋势结束的伪信号而被提早震出、反向操作**，因此在2015年后半年SSA策略净值波动剧烈。

## **2.3多窗口SSA趋势线组**

利用SSA相对MA更不易受到极端值影响的优势，同时为了最大程度紧跟市场趋势，我们以趋势线末端点的相对涨跌来判断趋势。具体的**在T时刻 SSAm(T)若大于（小于）SSAm(T-1) ，就认为依据此趋势线判断市场处于上涨（下跌）趋势。**

单一SSA趋势线仅考察单一周期下的市场趋势，判断不够全面。本报告的择时策略采用多窗口SSA趋势线组末端趋势共同判断市场方向。不同窗口长度的SSA考量的是价格序列在不同周期长度下的趋势变化，我们希望在大部分趋势线判断统一的情况下进行操作，从而相当于给趋势的判断增加一个过滤。

我们以五趋势线组（m=15,20,25,30,35）为例来对趋势进行判断，这里我们按照多数原则形成交易信号，即趋势线多数末端上涨即认为上涨，多数末端下跌即认为下跌。

```python
closeIndex = np.array(data['closeIndex']) # HS300指数收盘
length = len(closeIndex)
output={}
for m in [15,20,25,30,35]:
    t = 2*m
    key = 'SSA'+str(m)
    color_list = []
    score_output = []
    
    for i in range(70,length):      # 由于计算ssa35需要前70天数据，前70天不动作
        inputArray = closeIndex[:i]
        inputArray = inputArray[-t:]
        temp = SSA(inputArray,t,m) 
        if temp[-1]>=temp[-2]:
            score_output.append(1)   # 存放各ssa均线打分，便于统计最终看涨看跌
            color_list.append('|r')  # 颜色区分 
        else:
            score_output.append(-1)
            color_list.append('|g')
    output.update({key:(score_output,color_list)}) 
print len(output['SSA15'][0]),length
```

** 下图中红色表示趋势线看涨信号，绿色表示看跌信号。 从上到下依次为 SSA15,SSA20,SSA25,SSA30,SSA35**

```python
plt.figure(figsize=(25,5))
for i in range(len(output['SSA15'][1])):
    plt.plot(i,5, output['SSA15'][1][i], markersize=50) 
    plt.plot(i,4, output['SSA20'][1][i], markersize=50) 
    plt.plot(i,3, output['SSA25'][1][i], markersize=50) 
    plt.plot(i,2, output['SSA30'][1][i], markersize=50) 
    plt.plot(i,1, output['SSA35'][1][i], markersize=50) 
```

**多数原则处理后的看涨看跌信号**

```python
plt.figure(figsize=(25,5))
mwssa = []
for i in range(len(output['SSA15'][1])):
    score = output['SSA15'][0][i] + output['SSA20'][0][i] + output['SSA25'][0][i] + output['SSA30'][0][i] + output['SSA35'][0][i]
    if score>0:
        mark = '|r'
        mwssa.append(1)
        output.update()
    else:
        mark = '|g'
        mwssa.append(-1)
    plt.plot(i,0,mark,markersize=50)
    
```

下面计算**多数原则**的判断，与各条均线判断的重叠率（即看涨看跌相同）

```python
# 计算看涨看跌的重叠率
accuracy = [0,0,0,0,0]
num = len(mwssa)
for i in range(num):
    if mwssa[i] == output['SSA15'][0][i]:
        accuracy[0] += 1
    if mwssa[i] == output['SSA20'][0][i]:
        accuracy[1] += 1
    if mwssa[i] == output['SSA25'][0][i]:
        accuracy[2] += 1
    if mwssa[i] == output['SSA30'][0][i]:
        accuracy[3] += 1
    if mwssa[i] == output['SSA35'][0][i]:
        accuracy[4] += 1

for i in range(len(accuracy)):
    accuracy[i] = accuracy[i]/num
    print accuracy[i]
```

最终的判断结果在93.86%的时间里和**中枢趋势线**（SSA25）相同，高于其他趋势线（SSA15：85.66%，SSA20：90.53%，SSA30：92.39%，SSA35：89.07%）。因此对于趋势线组来说，最终的判断结果主要由其中枢趋势线决定，但由于其他趋势线信号的过滤作用，结果并不完全一样。

# **三、 基于奇异谱分析的择时策略**

均线择时策略属于动量策略，其策略构造往往涉及趋势的确认，趋势的跟随，趋势结束的确认和离场等过程，一般只适应趋势市而不适应震荡市。一个良好的均线择时策略应当能够在趋势市中抓住主要趋势，并在震荡市中保持稳定，避免过多的回撤。本节考察多窗口SSA趋势线组的择时效果。

## **3.1 参数选择**
多窗口SSA趋势线组作为一个趋势判断系统，其变量包含：

**1、趋势线的数量；**

**2、趋势线窗口参数选择。**

由于趋势线组对趋势的判断采用多数规则，因此趋势线组数量要求为**奇数**，从而方便策略设计。我们分别采用单趋势线，3趋势线组，5趋势线组进行回测。为了减小计算量，备选的趋势线窗口参数按等间隔取[5，10，15，20，25，30，35]。回测标的为沪深300指数，回测尺度为日线，回测区间为2005年1月到2016年4月，交易费用双边0.2%。

先确定参数的所有组合(排列组合问题。)

```python
import itertools # 带排列组合函数的包

# 选取趋势线参数组
parameter_dict = {}
for num_of_line in [1,3,5]:   # 趋势线数
    combinations = []
    options = itertools.combinations([5,10,15,20,25,30,35],num_of_line)
    for i in options:
        combinations.append(i)
    parameter_dict.update({num_of_line:combinations})

# 查看所有组合个数
for key in parameter_dict:
    print key,':',len( parameter_dict[key])
    #for comb in parameter_dict[key]:
        #print comb
```

## **3.2 整合回测函数及回测表现计算函数**

```python
from sklearn import linear_model

# 打包回测函数 
def LoopbackTest(data,combination):    
    ##############################################################        部分参数初始化          ####################################
    netValue = 1  #净值
    numOfTrading = 0 #交易次数
    if 'closeIndex' in data.columns:
        closeIndex = np.array(data['closeIndex']) # HS300指数收盘
        openIndex = np.array(data['openIndex'])
    elif 'closePrice' in data.columns:
        closeIndex = np.array(data['closePrice']) # HS300指数收盘
        openIndex = np.array(data['openPrice'])
    datelist  = np.array(data['tradeDate']) #日期
    flag = False                            # 持有标志，买入持仓则为True,否则为False
    strategy_name = 'strategy'+'('
    for i in range(len(combination)):
        strategy_name = strategy_name+str(combination[i])+','
    strategy_name = strategy_name[:-1]+')'
    strategy = {}          # 记录策略每天的净值
    length = len(closeIndex)  # 回测总天数 
    startHold =[]          #记录多头当期成本价及当期净值
    endHold = []
    t_max = 2*max(combination)  # 需要计算最大窗口ssa线 的最小天数
    #################################################################################################################################
    
    
    #$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$           每天的交易              $$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$
    # 假设每天开盘后以昨天的收盘价判断；但以今天的开盘价成交
    for i in range(t_max,length):     # 前t_max天不动作
        yesterday = datelist[i-1]
        today = datelist[i]
        valueYesterday = closeIndex[i-1]  #上个交易日的收盘价
        valueToday = openIndex[i]        # 若买入，开盘价为买入价
                
        # 计算当前净值(截止昨天收盘)
        if len(startHold)!=0:    #持有头寸
            if valueYesterday==0: # 停牌
                netValue = netValue
            else:
                netValue = (startHold[0])*(valueYesterday/startHold[1])  # 这里是计算当前值(昨天收盘价)与买入时成本的变化
        else: #空仓时，说明至少在昨天早上就清了仓，以昨天开盘价计算净值(卖出操作以计算);若在更早清仓，则净值就等于前天净值
            netValue = netValue
        strategy.update({yesterday:netValue})
        
        SSA_score = []    # 存放每个均线的看涨看跌，看涨为1，看跌为-1
        for m in combination:
            t = 2*m
            inputArray = closeIndex[:i]  # 取今天以前的数据
            inputArray = inputArray[-t:]
            SSAx = SSA(inputArray,t,m) 
            if SSAx[-1]>=SSAx[-2]:   # 判断看涨看跌
                SSA_score.append(1)
            else:
                SSA_score.append(-1)       
        score = sum(SSA_score)    # 各均线进行投票（求和），为正则看涨，为负看跌
        
        # 判断买入 
        if flag==False and score>0 and valueToday!=0:
            flag = True
            netValue = netValue * 0.998  # 扣除交易费用0.2%
            numOfTrading += 1  # 交易次数加1
            startHold = [netValue,valueToday]
    
        # 判断卖出
        if flag==True and score<0:
            flag = False
            netValue = (startHold[0])*(valueToday/startHold[1])  #以开盘价作为卖出价计算当前净值
            netValue = netValue * 0.998  # 扣除交易费用0.2%
            numOfTrading += 1  # 交易次数加1
            startHold = [] # 清仓
    #$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$
    return {strategy_name:(strategy,numOfTrading)}

# 年化利率
def AnnualizedReturns(pStart,pEnd,n):  #pStart为初始净值，pEnd为最终净值，n为策略持续时间（交易日）
    return (pEnd/pStart)**(250/n) -1

# 最大回撤（输入为list或array）
def MaxDrawdown(strategy):
    length = len(strategy)
    drawdown = []
    py =strategy[0]
    for i in range(1,length):
        px = strategy[i]
        py = max(strategy[:i])
        drawdown.append(1-(px/py))
    return max(drawdown)

# 斜率（输入为list或array）
def Slope(strategy):  
    num=len(strategy)
    x = np.mat(np.arange(num))
    x = x.T
    y = strategy - 1  # 截距为1
    clf = linear_model.LinearRegression()
    clf.fit(x,y)
    slope = clf.coef_[0]
    return slope

# 策略表现
def Performance(strategy):
    strategyName = strategy.keys()[0]
    netValueDict = strategy[strategyName][0]
    netValueArray = np.array(pd.Series(strategy[strategyName][0])) # 净值序列
    numOfTrading = strategy[strategyName][1]
    pStart = 1
    pEnd = netValueArray[-1] #最终净值
    n = len(netValueArray) # 交易日总数
    slope = Slope(netValueArray)  # 拟合斜率
    maxdrawdown = MaxDrawdown(netValueArray) #最大回撤
    annReturns = AnnualizedReturns(pStart,pEnd,n)
    performance = {}
    performance.update({u'拟合斜率':slope})
    performance.update({u'最终净值':pEnd})
    performance.update({u'年化收益率':annReturns})
    performance.update({u'最大回撤':maxdrawdown})
    performance.update({u'交易次数':numOfTrading})
    return {strategyName:performance}

# 自动测试，并生成表格
def AutoTest(data,parameter_dict):
    ret = {}
    for num_of_line in parameter_dict.keys():
        combinations = parameter_dict[num_of_line]
        performanceDF = pd.DataFrame()
        for combination in combinations:
            tempStrategy = LoopbackTest(data,combination)
            performance = Performance(tempStrategy)
            key = performance.keys()[0]
            performanceDF[key] = pd.Series(performance[key])
        ret.update({num_of_line:performanceDF.transpose()})
        print num_of_line,'条均线组合测试完成！'
    return ret
        
```
