```python
#导入软件包
import numpy as np
import pandas as pd
import scipy.stats as st
import datetime
import time 
import copy
import matplotlib.pyplot as plt
import statsmodels.api as sm
from CAL.PyCAL import *
universe = set_universe('A')
cal = Calendar('China.SSE')
```

```python
#定义时间转化函数
def time_change(x):
    y = datetime.datetime.strptime(x,"%Y-%m-%d")
    y = y.strftime("%Y%m%d")
    return y

#定义函数：获取研究时间段的交易日、每个月每月，每个月月初
def get_trade_list(start_date,end_date):
    cal_dates = DataAPI.TradeCalGet(exchangeCD=u"XSHG",beginDate=start_date,endDate=end_date,\
                                    field=u"calendarDate,isOpen,isMonthEnd",pandas="1").sort('calendarDate')
    cal_dates = cal_dates[cal_dates['isOpen']==1]
    
    #交易日历
    trade_list = cal_dates['calendarDate'].values.tolist()
    trade_list = [time_change(x) for x in trade_list]

    #每个月的月末
    month_end = cal_dates[cal_dates['isMonthEnd']==1]['calendarDate'].values.tolist()
    month_end = [time_change(x) for x in month_end]
    
    #下一个月的月初第一个交易日
    cal = Calendar('China.SSE')
    month_start = [cal.advanceDate(x,'1B').strftime("%Y%m%d") for x in month_end]
    
    return trade_list,month_end,month_start

#get_trade_list('20160723','20170925')

#定义剔除ST股票的函数  source_universe有后缀为list,st_date = None默认为当天，返回有后缀的代码
def st_remove(source_universe,st_date = None):
    st_date = st_date if st_date is not None else datetime.datetime.now().strftime("%Y%m%d")
    df_ST = DataAPI.SecSTGet(beginDate=st_date,endDate=st_date,secID=source_universe,field=u"secID")['secID'].values.tolist()
    return [x for x in source_universe if x not in df_ST]

#剔除新上市的股票（认为上市不满N天的为新上市股票）ticker无后缀，day为int
def new_remove(ticker,tradeDate = None,day = 1):
    tradeDate = tradeDate if tradeDate is not None else datetime.datetime.now().strftime("%Y%m%d")
    period = '-'+str(day)+'B'
    pastDate = cal.advanceDate(tradeDate,period).strftime("%Y-%m-%d")
    ipo_date = DataAPI.SecIDGet(ticker=ticker,assetClass=u"E",field=u"ticker,listDate",pandas="1")
    end_ticker=ipo_date[ipo_date['listDate']<pastDate]['ticker'].tolist()
    return end_ticker

#定义函数，将无后缀股票代码（list）转化为平台带有后缀的股票代码（list）
def ticker2secID(ticker):
    universe_1 = DataAPI.EquGet(equTypeCD=u"A",listStatusCD=u"L,S,DE,UN",field=u"ticker,secID",pandas="1")
    universe_1 = dict(universe_1.set_index('ticker')['secID'])
    if isinstance(ticker,list):
        res = []
        for i in ticker:
            if i in list(universe_1.keys()):
                 res.append(universe_1[i])            
            else:
                 print(i+'在universe_1中没有找到')
        return res
    else:
        print('ticker should be list')

#获取每月月初的所有股票池DF,date月初的时间，N为new_remove函数中的N，返回DF
def get_stock_pool(date,N):
    univ = DynamicUniverse('A')
    all_code = univ.preview(date,skip_halted=False)
    all_code_not_ST = st_remove(all_code,st_date = date)
    ticker = [x.split('.')[0] for x in all_code_not_ST]
    all_code_need = new_remove(ticker,tradeDate =date,day = N)
    code = ticker2secID(all_code_need)
    df = pd.DataFrame({'code':code})
    df['date']=date
    df = df[['date','code']]
    return df
#get_stock_pool('20100301',100) 1 归纳总结小的板块   2 理解 转化为自己的，运用  3 细节 认真  4 方法
```

```python
#获取20071220-2018001期间，每个月月初的所有A股股票的代码集合
start_date = '20071220'
end_date = '20100601'
trade_list,month_end,month_start = get_trade_list(start_date,end_date)
N = 180
all_stock = pd.DataFrame()
for date in month_start:
    stock = get_stock_pool(date,N)
    all_stock = pd.concat([all_stock,stock],axis = 0)
all_stock.head(5)
```

```python
#获取所有股票代码的集合
all_code =sorted(list(set(all_stock['code'])))

#获取20071120-20180601期间所有的每天交易数据
mkt_info_frame = DataAPI.MktEqudGet(secID=all_code,beginDate=u"20071120",endDate=u"20100601",\
                                    field=u"secID,tradeDate,chgPct,turnoverRate",pandas="1")
mkt_info_frame['tradeDate'] = map(time_change,mkt_info_frame['tradeDate'])
mkt_info_frame.columns = ['code','date','ret','to']
mkt_info_frame=mkt_info_frame.dropna()
mkt_info_frame.head(5)
```

```python
#构造因子FR,FRu,FRd
all_signal= pd.DataFrame()  #用于储存因子以及下一个月的收益
#for i in range(1,2):
for i in range(1,len(month_end)):
    begin = month_start[i-1]
    end = month_end[i]
    start= month_start[i]
    if i < len(month_end) - 1:
        next_end = month_end[i + 1]
    else:
        next_end = end
    #次月的收益
    code =all_stock[all_stock['date']==start]['code'].tolist()
    close1 = DataAPI.MktEqudAdjGet(secID=code,tradeDate=end,\
                                   field=u"secID,closePrice",pandas="1")
    close1.columns = ['code','close1']
    close2 = DataAPI.MktEqudAdjGet(secID=code,tradeDate=next_end,\
                                   field=u"secID,closePrice",pandas="1")
    close2.columns = ['code','close2']
    
    close = pd.merge(close1,close2,on ='code',how ='outer')
    close['month_ret']=close['close2']/close['close1']-1
    close['date']=end
    close = close[['code','date','month_ret']]

       
    #因子
    period_frame =mkt_info_frame[(mkt_info_frame['date']>=begin)&(mkt_info_frame['date']<=end)] 
    period_frame = period_frame[period_frame['code'].isin(list(all_stock[all_stock['date']==start]['code']))]
    period_frame['ret_next']=period_frame.groupby('code')['ret'].shift(-1)
    period_frame['abs_ret_next']=abs(period_frame['ret_next'])
    period_frame.dropna(inplace=True)
    
    trading_day = [x for x in trade_list if x>=begin and x<end]
    n_days = int(len(trading_day)/2)
    
    #FR
    signal_fr = period_frame.groupby('code').apply(lambda df:df['to'].corr(df['abs_ret_next'],method = 'pearson',min_periods = n_days))
    
    #FRu
    period_frame_positive = period_frame[period_frame['ret_next']>0]
    signal_fru = period_frame_positive.groupby('code').apply(lambda df:df['to'].corr(df['ret_next'],method = 'pearson',min_periods = 5))
    
    #FRd
    period_frame_negative = period_frame[period_frame['ret_next']<0]
    signal_frd = period_frame_negative.groupby('code').apply(lambda df:df['to'].corr(df['ret_next'],method = 'pearson',min_periods = 5))
    
    #整合
    signal = pd.concat([signal_fr,signal_fru],axis=1,join = 'outer')
    signal = pd.concat([signal,signal_frd],axis=1,join = 'outer')
    
    signal.columns=['fr','fru','frd']
    signal['code']=signal.index
    signal['date']=end
    
    signal = signal[['date','code','fr','fru','frd']]
    signal.sort_values(by = 'code',inplace = True)
    signal.reset_index(drop = True,inplace = True)

    signal = pd.merge(signal,close,how ='right')
    
    all_signal = all_signal.append(signal)
    
signal
```
