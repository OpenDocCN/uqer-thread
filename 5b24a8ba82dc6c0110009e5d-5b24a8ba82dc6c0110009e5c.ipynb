```python
import pandas as pd
import numpy as np
from CAL.PyCAL import *
import datetime
import talib

start = '2016-01-01'                  
end = '2018-06-01'                  
universe = ['IM0']#['ICM0','IFM0','IHM0','TFM0','TM0','RBM0','IM0','ZNM0','MM0','PM0','RUM0','AUM0','CFM0','FGM0','NIM0','RMM0','JM0','YM0','AGM0','BUM0','LM0','PPM0','SRM0','TAM0','TCM0','AM0','CM0','HCM0','JMM0','MAM0','OIM0','CSM0','JDM0']      
benchmark = 'HS300'
freq = 'd'                                 
refresh_rate = 1                           
cal=Calendar('China.SSE')

#导入weekly期货交易数据
comp = dict()
scenario = dict()
for future in universe:
    df = DataAPI.MktFutwGet(ticker=[future[:-2] + str(m) + n for m in range(1,19) for n in ['01','02','03','04','05','06','07','08','09','10','11','12']],beginDate=u"20130101",endDate=u"20180601",field=['ticker','endDate','openPrice','highestPrice','lowestPrice','closePrice','turnoverVol','openInt'],pandas="1")
    df = df.sort(['endDate','openInt'], ascending = [1,0])
    df = df.drop_duplicates('endDate')
    df.set_index('endDate',inplace=True)
    
    #计算MACD
    macd_tmp = talib.MACD(np.array(df['closePrice'],dtype='f8'), fastperiod=12, slowperiod=26, signalperiod=9)
    df['DIF'] = macd_tmp[0]
    df['DIF_change'] = df['DIF'] - df['DIF'].shift(1)
    df['variance'] = df['closePrice'].rolling(20).std()
    df['DIF_pct'] = df['DIF_change'] / df['variance']
    comp[future] = df
    
accounts = {
    'futures_account': AccountConfig(account_type='futures', capital_base=30000)
}


#quartz参数设置
sim_params = quartz.SimulationParameters(start, end, benchmark, universe,
                                         freq=freq,
                                         refresh_rate=refresh_rate,
                                         accounts=accounts)
data = quartz.get_backtest_data(sim_params)
results = {}

#策略主体
for thres in [0.001, 0.005, 0.01]: 
    def initialize(context):
        pass

    def handle_data(context):    
        today = context.current_date
        yesterday = cal.advanceDate(today,'-1B',BizDayConvention.Preceding)
        today = today.strftime('%Y-%m-%d')
        yesterday = yesterday.strftime('%Y-%m-%d')
        futures_account = context.get_account('futures_account')
        current_universe = context.get_universe('futures')
        mktvalue = futures_account.portfolio_value

        for future in current_universe:
            symbol = context.get_symbol(future)
            position = futures_account.get_positions().get(symbol, dict())
            long_position = position.get('long_amount', 0)
            short_position = position.get('short_amount', 0)

            if context.mapping_changed(future):
                symbol_before, symbol_after = context.get_rolling_tuple(future)
                position = futures_account.get_positions().get(symbol_before, dict())
                long_position = position.get('long_amount', 0)
                short_position = position.get('short_amount', 0)
                if futures_account.get_position(symbol_before):
                    futures_account.switch_position(symbol_before, symbol_after)

            if yesterday in comp[future].index:
                df = comp[future]
                row_name = df[df.index == yesterday].iloc[-1]
                row_num = df.index.get_loc(row_name.name)
                DIF_pct = df.ix[row_num,'DIF_pct']

                if long_position == short_position == 0 :
                    if DIF_pct > thres:
                        futures_account.order(symbol, 1, 'open')
                    if DIF_pct < -thres:
                        futures_account.order(symbol, -1, 'open')

                elif long_position > 0 :
                    if DIF_pct < -thres:
                        futures_account.order(symbol, -1, 'close')
                        futures_account.order(symbol, -1, 'open')
                elif short_position > 0 :
                    if DIF_pct > thres:
                        futures_account.order(symbol, 1, 'close')
                        futures_account.order(symbol, 1, 'open')

    #运行策略并记录
    strategy = quartz.TradingStrategy(initialize, handle_data)
    bt, perf = quartz.quick_backtest(sim_params, strategy, data=data)
    results[thres] = {'annualized_return': perf['annualized_return'],
                      'volatility': perf['volatility'],
                      'max_drawdown': perf['max_drawdown']
                      }
    
results = pd.DataFrame(results)
results
```
