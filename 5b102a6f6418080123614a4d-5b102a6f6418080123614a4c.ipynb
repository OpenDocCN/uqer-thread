# 三层因子检测模型-长期因子

```python
import pandas as pd
import numpy as np
import datetime
import time
```

```python
today = time.strftime("%Y%m%d")
```

```python
## 获取当前所有股票代码
today_tick = DataAPI.MktEqudGet(secID=u"",ticker=u"",tradeDate=today,beginDate=u"",endDate=u"",isOpen="",field=u"ticker",pandas="1")
today_tick = list(today_tick['ticker'].values)
```

```python
## 盈利指标
yl_index = DataAPI.FdmtIndiRtnPitGet(ticker=today_tick,secID="",endDate="",beginDate="",beginYear=u"",endYear=u"",reportType=['S1', 'A'],publishDateEnd=u"",publishDateBegin=u"",field=u"",pandas="1")
## 个股指标
gg_index = DataAPI.FdmtIndiPSPitGet(ticker=today_tick,secID="",endDate="",beginDate="",beginYear=u"",endYear=u"",reportType=['S1', 'A'],publishDateEnd=u"",publishDateBegin=u"",field=u"",pandas="1")
## 成长指标
cz_index = DataAPI.FdmtIndiGrowthPitGet(ticker=today_tick,secID="",endDate="",beginDate="",beginYear=u"",endYear=u"",reportType=['S1', 'A'],publishDateEnd=u"",publishDateBegin=u"",field=u"",pandas="1")
## 运营能力
yy_index = DataAPI.FdmtIndiTrnovrPitGet(ticker=today_tick,secID="",endDate="",beginDate="",beginYear=u"",endYear=u"",reportType=['S1', 'A'],publishDateEnd=u"",publishDateBegin=u"",field=u"",pandas="1")
```

```python
def pre_process_data(df, date = '2010-01-01'):
    new_df = df.copy()
    ## 取2010年之后数据
    new_df = new_df[new_df['publishDate'] > date]
    ## 去重值
    df_t = new_df[['ticker', 'endDate']]
    new_df = new_df.loc[df_t.drop_duplicates(keep = 'last').index, :]
    return new_df
```

```python
## 对数据进行处理
yl_index = pre_process_data(yl_index)
gg_index = pre_process_data(gg_index)
cz_index = pre_process_data(cz_index)
yy_index = pre_process_data(yy_index)
```

```python
## 计算财报数据发布以来的前一时间段
def drop_value(df):
    new_df = df.copy()
    new_df['PrePublishDate'] = new_df.groupby('ticker')['publishDate'].apply(lambda x: x.shift(-1))
    new_df = new_df.loc[new_df[['publishDate', 'PrePublishDate']].dropna().index,:]
    return new_df
```

```python
## 计算自财报发出之后到下一次财报发布期间的表现
def get_ratio(df):
    flag = 0
    new_df = drop_value(df)
    ratio_list = []
    for i in new_df[['ticker','publishDate', 'PrePublishDate']].values:
        flag += 1
        ticker, date_now, date_pre = i
        try:
            df_price = DataAPI.MktEqudGet(ticker=ticker , beginDate=date_pre ,endDate=date_now, field=['closePrice'],pandas="1")['closePrice']
            ratio = df_price.iloc[-1] - df_price.iloc[0] /  df_price.iloc[0]
            ratio_list.append(ratio)
        except:
            ratio_list.append(np.nan)
        if flag % 5000 == 0:
            print(flag, 'has done')
    new_df['ratio'] = ratio_list
    return new_df
```

```python
yl_ratio = get_ratio(yl_index)
gg_ratio = get_ratio(gg_index)
cz_ratio = get_ratio(cz_index)
yy_ratio = get_ratio(yy_index)
```

```python
def fitter_col(k):
    key_words = ['ID', 'Date', 'time','Name', 'exchangeCD']
    f_1 = [(i in k) for i in key_words]
    if sum(f_1) > 0 :
        return False
    else:
        return True
```

```python
from scipy.stats import spearmanr
```

```python
def get_effect_fac(df):
    df_copy = df.copy()
    ## 年份数据
    df_copy['year'] = df_copy['publishDate'].apply(lambda x: x[:4])
    fac_list = [i for i in df_copy.columns if fitter_col(i)]
    yl_g_df = df_copy.groupby(['year', 'ticker'])[fac_list].apply(lambda x: x[:])
    ## 计算年份数据与变量之间的相关性
    new_df = pd.DataFrame()
    for j in yl_g_df.index.levels[0]:
        for i in [k for k in fac_list if k not in ['year', 'ticker', 'ratio']]:
            y, x = yl_g_df.loc[j][['ratio', i]].dropna().values.T
            if spearmanr(y, x)[1] < 0.5:
                new_df.loc[j, i] = spearmanr(y,x)[0]
            else:
                new_df.loc[j, i] = np.nan
    return new_df
```

```python
yl_sp_df = get_effect_fac(yl_ratio)
gg_sp_df = get_effect_fac(gg_ratio)
cz_sp_df = get_effect_fac(cz_ratio)
yy_sp_df = get_effect_fac(yy_ratio)
```
