### 获取数据

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import seaborn as sns
from pandas import DataFrame,Series
import matplotlib.pyplot as plt
```

### 协整关系的检验

我们想使用协整的特性进行配对交易，那么要怎么样发现协整关系呢？在 Python 的 Statsmodels 包中，有直接用于协整关系检验的函数 coint，该函数包含于 statsmodels.tsa.stattools 中。首先，我们构造一个读取品种价格，判断协整关系的函数。该函数返回的两个值分别为协整性检验的 p 值矩阵以及所有传入的参数中协整性较强的品种。 p 值越低，协整关系就越强；p 值低于 0.050.05 时，协整关系便非常强。

```python
def find_cointegrated_pairs(dataframe):
    n = dataframe.shape[1]
    pvalue_matrix = np.ones((n, n))
    keys = dataframe.keys()
    pairs = []
    for i in range(n):
        for j in range(i+1, n):
            stock1 = dataframe[keys[i]]
            stock2 = dataframe[keys[j]]
            result = sm.tsa.stattools.coint(stock1, stock2)
            pvalue = result[1]
            pvalue_matrix[i, j] = pvalue
            if pvalue < 0.05:
                pairs.append((keys[i], keys[j], pvalue))
    return pvalue_matrix, pairs
```

```python
con = list(set(data['contractObject']))
df = DataFrame()
for i in con:
    df2 = DataFrame(data[data['contractObject']==i]['settlePrice'])
    df2.columns = [i]
    df = pd.concat([df,df2], axis=1)
df.dropna().head()
```

```python
pvalues, pairs = find_cointegrated_pairs(df)
sns.heatmap(1-pvalues, xticklabels=df.columns, yticklabels=df.columns, cmap='RdYlGn_r', mask = (pvalues == 1))
p = DataFrame(pairs,columns=['S1','S2','Pvalue'])
p.sort_index(by='Pvalue')
```

```python
stock_df1 = df['J']
stock_df2 = df['JM']
stock_df1.plot(); stock_df2.plot()
plt.xlabel("Time"); plt.ylabel("Price")
plt.legend(["J", "JM"],loc='best')
```

接下来，我们用(CU	MA)的价格来进行一次OLS线性回归，以此算出它们是以什么线性组合的系数构成平稳序列的。

```python
x = stock_df1
y = stock_df2
X = sm.add_constant(x)
result = (sm.OLS(y,X)).fit()
print(result.summary())
```

```python
fig, ax = plt.subplots(figsize=(8,6))
ax.plot(x, y, 'o', label="data")
ax.plot(x, result.fittedvalues, 'r', label="OLS")
ax.legend(loc='best')
```

画出它们价差的平稳序列。可以看出，虽然价差上下波动，但都会回归中间的均值。
