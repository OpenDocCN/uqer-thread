```python
#阅读：一个简单的策略，利用指数10年7倍 
#http://mp.weixin.qq.com/s?__biz=MzA4ODg5NDk1OA==&mid=402742311&idx=2&sn=877dca110bae18b5ae9ed4930ab955a7#rd
#要点：选择上证指数，指数如果创10天新高就买入，跌破10天新低就卖出。

#分析目标：
#标的：指数，SHS300，ZZ500，CYB
#时间:20130101-20140701,20140701-20150615,20150615-20160101

import numpy as np
import pandas as pd
import CAL.PyCAL as pycal
import matplotlib.pyplot as plt

#取得[]中近n天最大的元素，method=1最大，method=-1最小
def getMaxList(tmpList,n,method=1):
    resultList=[]
    if len(tmpList)==0:
        raise Exception("Except parameter empty")
    maxValue=tmpList[0]
    if method==1:
        for i in xrange(0,len(tmpList)):
            if tmpList[i] > maxValue:
                maxValue=tmpList[i]
            else:
                if i>n and tmpList[i-n-1]==maxValue:
                    maxValue=tmpList[i-n:i].max()
            resultList.append(maxValue)
    else:
        for i in xrange(0,len(tmpList)):
            if tmpList[i] < maxValue:
                maxValue=tmpList[i]
            else:
                if i>n and tmpList[i-n-1]==maxValue:
                    maxValue=tmpList[i-n:i].min()
            resultList.append(maxValue)
    return resultList
        
    
fmt='{:.4f}'
start='20130101'
end='20160130'
dateList=['20050101','20150501']#'20130101','20140701','20150615','20160101'
all_stock_prices=pd.DataFrame()
print '### start '
stock_prices=None
fields=['secID','tradeDate','closePrice','accumAdjFactor','highestPrice','lowestPrice']
fieldsIndex=['ticker','tradeDate','closeIndex','highestIndex','lowestIndex','openIndex']
resultsTable=pd.DataFrame()
resultsTable=resultsTable.append(['date','stockSecID','profit'])
for i in xrange(0,len(dateList)-1):
    start=dateList[i]
    end=dateList[i+1]
    for stockSecID in ['399006','000016','000001']:
		#获取价格数据
        stock_prices=None
        try:  
            stock_prices=DataAPI.MktIdxdGet(indexID=u"",ticker=stockSecID,tradeDate='',beginDate=start,endDate=end,field=fieldsIndex,pandas="1")
            stock_prices.rename(columns={'ticker':'stockSecID','closeIndex':'closePrice','highestIndex':'highestPrice',\
                                         'lowestIndex':'lowestPrice','openIndex':'openPrice'}, inplace=True)
        except Exception ,e:
            print e
        if stock_prices is None or stock_prices.size==0:
            print 'stock_prices is null:',start,end,stockSecID
            continue
		#计算列-10日最高最低价
        stock_prices['10MaxPrice']=getMaxList(stock_prices['closePrice'].values,10,1)
        stock_prices['10MinPrice']=getMaxList(stock_prices['closePrice'].values,10,-1)
 
        
      #计算列-买入卖出信号序列
        stock_prices['signal']=0
        stock_prices['buySignal']=(stock_prices['10MaxPrice']==stock_prices['closePrice']).apply(lambda x:1 if x==True else 0)
        stock_prices['sellSignal']=(stock_prices['10MinPrice']==stock_prices['closePrice']).apply(lambda x:-1 if x==True else 0)
        stock_prices['signal']=stock_prices['buySignal'].apply(lambda x:None if x==0 else x)
        stock_prices['signal']=stock_prices['signal'].combine_first(stock_prices['sellSignal'])
        
        #print '##',stock_prices[['tradeDate','closePrice','10MaxPrice','10MinPrice','signal']][0:25]

      #买入卖出信号连续化，消除，买入，买入等重复信号
        nowStatus=0
        for i in xrange(0,stock_prices.shape[0]):
            if stock_prices['signal'][i]==-1 and nowStatus==0:#无持仓要卖
                stock_prices['signal'][i]=0
            elif stock_prices['signal'][i]==-1 and nowStatus==1:#有持仓要卖
                nowStatus=0
            elif stock_prices['signal'][i]==1 and nowStatus==0:#无持仓要买
                nowStatus=1
            elif stock_prices['signal'][i]==1 and nowStatus==1:#有持仓要买
                stock_prices['signal'][i]=0
        
        #计算收益-信号序列计算收益（买入卖出拼成一条记录，然后计算涨跌率，之后cumprod
        stock_prices=stock_prices[stock_prices['signal']!=0]
        stock_prices['nextTradeDate']=np.append(np.array(stock_prices['tradeDate'][1:]),None)
        stock_prices['nextClosePrice']=np.append(np.array(stock_prices['closePrice'][1:]),None)
        stock_prices['nextSignal']=np.append(np.array(stock_prices['signal'][1:]),None)
        stock_prices=stock_prices[np.logical_and(stock_prices['signal']==1,stock_prices['nextSignal']==-1)]
        #print '###',stock_prices[['tradeDate','closePrice','nextTradeDate','nextClosePrice','signal','nextSignal']][0:25]
        
        stock_prices['profitRate']=stock_prices['nextClosePrice']/stock_prices['closePrice']

        stock_prices['profitRateSum']=stock_prices['profitRate'].cumprod()
        #print stock_prices[['profitRateSum','profitRate']]
        print start,end,stockSecID,stock_prices['profitRateSum'].values[-1]
print '### end '
```
