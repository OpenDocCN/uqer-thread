# 

看财经新闻，经常看到“黑天鹅指数创历史新高”这样的标题，这其中的黑天鹅指数，一般是指芝加哥期权交易所（CBOE）推出的偏度指数(CBOE SKEW Index)，该指数通过标普500指数价外期权价格的偏离程度来衡量市场对意外事件的担忧程度。目前A股有50ETF期权，所以同样的，我们可以构建对应于上证50指数的黑天鹅指数。

# 1. 偏度简介

---

在概率论和统计学中，偏度衡量随机变量概率分布的不对称性。偏度的值可以为正，可以为负。在数量上，偏度为负（负偏态）就意味着在概率密度函数左侧的尾部比右侧的长，绝大多数的值（包括中位数在内）位于平均值的右侧。偏度为正（正偏态）就意味着在概率密度函数右侧的尾部比左侧的长，绝大多数的值（但不一定包括中位数）位于平均值的左侧。偏度为零就表示数值相对均匀地分布在平均值的两侧，但不一定意味着其为对称分布。


- 负偏态或左偏态：**左侧的尾部更长**，分布的主体集中在右侧
- 正偏态或右偏态：**右侧的尾部更长**，分布的主体集中在左侧

---

下图为标普500月度对数收益率的分布图，可以很明显地看出来有厚尾特征：

- 红色柱状分布图里面，相对较大跌幅发生的概率明显比正态分布要大一点

![](http://ww3.sinaimg.cn/large/52041eb9gw1f6app2qq7nj20m80fbgmq.jpg 'dist')

---

偏度在数学上的定义也很简单：
$$S = E\bigg[\bigg(\frac{R-\mu}{\sigma}\bigg)^3\bigg]$$
其中R表示月度对数收益，$\mu$是该收益序列的平均值，$\sigma$是其标准差，E表示数学期望。

如上计算出来的偏度S其值一般较小，为了方便，我们做个变换：
$$\mathrm{SKEW} = 100 - 10\times S$$
经过这一变换，如果尾部风险加大的话，SKEW就会增大。另外，这样定义的SKEW其值为100时代表基本无偏度。

# 2. SKEW 黑天鹅指数计算

---

本节中，我们讨论从50ETF期权市场行情中，如何估算出期权交易者眼中的尾部风险，即通过期权市场行情计算SKEW。

对不同行权价的期权，计算其隐含波动率，会呈现两头略高中间略低的形态，称之为隐含波动率微笑。下图为标普在1987年大跌之前和之后的隐含波动率微笑示意图：

![](http://ww1.sinaimg.cn/large/52041eb9gw1f6apolu3ojj20m80d1wes.jpg 'smile')

红色实线在深度虚值看跌期权这头的上翘表示出期权交易者对于尾部风险的担忧，而通过期权市场的行情，我们可以提取出其中隐含的SKEW偏度数据。

----

具体地计算方法不再赘述，和计算中国波指iVIX的方法非常类似，细节可以参考CBOE的skew白皮书：
http://www.cboe.com/micro/skew/documents/skewwhitepaperjan2011.pdf

需要注意的是，**此处我们计算得到的SKEW，是期权市场交易行情隐含的偏度，并不是通过现货市场的每日收益数据计算得到**

```python
from CAL.PyCAL import * 
import pandas as pd
import numpy as np
from scipy import interpolate
import matplotlib.pylab as plt
```

```python
## 银行间质押式回购利率
def histDayInterestRateInterbankRepo(date):
    cal = Calendar('China.SSE')
    period = Period('-10B')
    begin = cal.advanceDate(date, period)
    begin_str = begin.toISO().replace('-', '')
    date_str = date.toISO().replace('-', '')
    # 以下的indicID分别对应的银行间质押式回购利率周期为：
    # 1D, 7D, 14D, 21D, 1M, 3M, 4M, 6M, 9M, 1Y
    indicID = [u"M120000067", u"M120000068", u"M120000069", u"M120000070", u"M120000071", 
               u"M120000072", u"M120000073", u"M120000074", u"M120000075", u"M120000076"]
    period = np.asarray([1.0, 7.0, 14.0, 21.0, 30.0, 90.0, 120.0, 180.0, 270.0, 360.0]) / 360.0
    period_matrix = pd.DataFrame(index=indicID, data=period, columns=['period'])
    field = u"indicID,indicName,publishTime,periodDate,dataValue,unit"
    interbank_repo = DataAPI.ChinaDataInterestRateInterbankRepoGet(indicID=indicID,beginDate=begin_str,endDate=date_str,field=field,pandas="1")
    interbank_repo = interbank_repo.groupby('indicID').first()
    interbank_repo = pd.concat([interbank_repo, period_matrix], axis=1, join='inner').sort_index()
    return interbank_repo

## 银行间同业拆借利率
def histDaySHIBOR(date):
    cal = Calendar('China.SSE')
    period = Period('-10B')
    begin = cal.advanceDate(date, period)
    begin_str = begin.toISO().replace('-', '')
    date_str = date.toISO().replace('-', '')
    # 以下的indicID分别对应的SHIBOR周期为：
    # 1D, 7D, 14D, 1M, 3M, 6M, 9M, 1Y
    indicID = [u"M120000057", u"M120000058", u"M120000059", u"M120000060", 
               u"M120000061", u"M120000062", u"M120000063", u"M120000064"]
    period = np.asarray([1.0, 7.0, 14.0, 30.0, 90.0, 180.0, 270.0, 360.0]) / 360.0
    period_matrix = pd.DataFrame(index=indicID, data=period, columns=['period'])
    field = u"indicID,indicName,publishTime,periodDate,dataValue,unit"
    interest_shibor = DataAPI.ChinaDataInterestRateSHIBORGet(indicID=indicID,beginDate=begin_str,endDate=date_str,field=field,pandas="1")
    interest_shibor = interest_shibor.groupby('indicID').first()
    interest_shibor = pd.concat([interest_shibor, period_matrix], axis=1, join='inner').sort_index()
    return interest_shibor

## 插值得到给定的周期的无风险利率
def periodsSplineRiskFreeInterestRate(date, periods):
    # 此处使用SHIBOR来插值
    init_shibor = histDaySHIBOR(date)
    
    shibor = {}
    min_period = min(init_shibor.period.values)
    min_period = 25.0/360.0
    max_period = max(init_shibor.period.values)
    for p in periods.keys():
        tmp = periods[p]
        if periods[p] > max_period:
            tmp = max_period * 0.99999
        elif periods[p] < min_period:
            tmp = min_period * 1.00001
        sh = interpolate.spline(init_shibor.period.values, init_shibor.dataValue.values, [tmp], order=3)
        shibor[p] = sh[0]/100.0
    return shibor
```

```python
# 计算历史某一天的Skew Index
# Based on http://www.cboe.com/micro/skew/introduction.aspx
# http://www.cboe.com/micro/skew/documents/skewwhitepaperjan2011.pdf
def calDaySkewIndex(date, opt_info):    
    var_sec = u"510050.XSHG"
    # 使用DataAPI.MktOptdGet，拿到历史上某一天的期权行情信息
    date_str = date.toISO().replace('-', '')
    fields_mkt = [u"optID", "tradeDate", "closePrice", 'settlPrice']
    opt_mkt = DataAPI.MktOptdGet(tradeDate=date_str, field=fields_mkt, pandas="1")
    opt_mkt = opt_mkt.set_index(u"optID")
    opt_mkt[u"price"] = opt_mkt['closePrice']

    # concat某一日行情和期权基本信息，得到所需数据
    opt = pd.concat([opt_info, opt_mkt], axis=1, join='inner').sort_index()
    opt = opt[opt.varSecID==var_sec]
    exp_dates = map(Date.parseISO, np.sort(opt.expDate.unique()))
    trade_date = date
    exp_periods = {}
    for epd in exp_dates:
        exp_periods[epd] = (epd - date)*1.0/365.0
    risk_free = periodsSplineRiskFreeInterestRate(trade_date, exp_periods)

    s = {}
    for date in exp_dates:
        # 计算每个期限的skew
        opt_date = opt[opt.expDate==date.toISO()]
        rf = risk_free[date]
        # rf = 0.06
        
        opt_call = opt_date[opt_date.contractType == 'CO'].set_index('strikePrice')
        opt_put = opt_date[opt_date.contractType == 'PO'].set_index('strikePrice')
        opt_call_price = opt_call[[u'price']].sort_index()
        opt_put_price = opt_put[[u'price']].sort_index()
        opt_call_price.columns = [u'callPrice']
        opt_put_price.columns = [u'putPrice']
        opt_call_put_price = pd.concat([opt_call_price, opt_put_price], axis=1, join='inner').sort_index()
        opt_call_put_price['diffCallPut'] = opt_call_put_price.callPrice - opt_call_put_price.putPrice
        
        strike = abs(opt_call_put_price['diffCallPut']).idxmin()
        price_diff = opt_call_put_price['diffCallPut'][strike]
        ttm = exp_periods[date]
        fw = strike + np.exp(ttm*rf) * price_diff
        strikes = np.sort(opt_call_put_price.index.values)
        
        delta_K_tmp = np.concatenate((strikes, strikes[-1:], strikes[-1:])) 
        delta_K_tmp = delta_K_tmp - np.concatenate((strikes[0:1], strikes[0:1], strikes))
        delta_K = np.concatenate((delta_K_tmp[1:2], delta_K_tmp[2:-2]/2, delta_K_tmp[-2:-1]))
        delta_K = pd.DataFrame(delta_K, index=strikes, columns=['deltaStrike'])
        
        # opt_otm = opt_out_of_money
        opt_otm = pd.concat([opt_call[opt_call.index>fw], opt_put[opt_put.index<fw]], axis=0, join='inner')
        opt_otm = pd.concat([opt_otm, delta_K], axis=1, join='inner').sort_index()
        
        # 计算skew时，比forward price低的第一个行权价被设置为参考行权价，参考值以上
        # 的call和以下的put均为虚值期权，所有的虚值期权被用来计算skew，然而计算中发
        # 现，有时候没有比forward price更低的行权价，例如2015-07-08，故有以下关于
        # 参考行权价的设置
        strike_ref = fw
        if len((strikes[strikes < fw])) > 0:
            strike_ref = max([k for k in strikes[strikes < fw]])
            opt_otm['price'][strike_ref] = (opt_call['price'][strike_ref] + opt_call['price'][strike_ref])/2.0
        
        exp_rt = np.exp(rf*ttm)
        ki = opt_otm.index
        qki = opt_otm.price
        deltaki = opt_otm.deltaStrike
        
        opt_otm['pp1'] = - qki*deltaki/ki/ki
        opt_otm['pp2'] = 2.0*(1.0 - np.log(ki/fw))*qki*deltaki/ki/ki
        opt_otm['pp3'] = 3.0*(2.0*np.log(ki/fw)-np.log(ki/fw)*np.log(ki/fw))*qki*deltaki/ki/ki
        ln2fk = np.log(fw/strike_ref)*np.log(fw/strike_ref)
        epsilon1 = -(1.0+np.log(fw/strike_ref)-fw/strike_ref)
        epsilon2 = 2.0*np.log(fw/strike_ref)*(fw/strike_ref-1)+0.5*ln2fk
        epsilon3 = ln2fk*(np.log(fw/strike_ref)-3+3*fw/strike_ref)
        
        p1 = exp_rt*sum(opt_otm['pp1']) + epsilon1
        p2 = exp_rt*sum(opt_otm['pp2']) + epsilon2
        p3 = exp_rt*sum(opt_otm['pp3']) + epsilon3
        
        numerator = p3 - 3*p1*p2 + 2*p1*p1*p1
        denominator = np.power(p2-p1*p1, 1.5)
        s[date] = numerator/denominator
    
    # d_one, d_two 将被用来计算SKEW(30):
    if exp_periods[exp_dates[0]] >= 1.0/365.0:
        d_one = exp_dates[0]
        d_two = exp_dates[1]
    else:
        d_one = exp_dates[1]
        d_two = exp_dates[2]
    w = (exp_periods[d_two] - 30.0/365.0)/(exp_periods[d_two] - exp_periods[d_one])
    skew30 = w*s[d_one] + (1 - w)*s[d_two]
    skew30 = 100 - 10*skew30
    
    # d_one, d_two 将被用来计算SKEW(60):
    d_one = exp_dates[1]
    d_two = exp_dates[2]
    w = (exp_periods[d_two] - 60.0/365.0)/(exp_periods[d_two] - exp_periods[d_one])
    skew60 = w*s[d_one] + (1 - w)*s[d_two]
    skew60 = 100 - 10*skew60
    
    return skew30,skew60

def getHistDailySkewIndex(beginDate, endDate):
    # 计算历史一段时间内的Skew Index指数并返回
    optionVarSecID = u"510050.XSHG"
    
    # 使用DataAPI.OptGet，一次拿取所有存在过的期权信息，以备后用
    fields_info = ["optID", u"varSecID", u'contractType', u'strikePrice', u'expDate']
    opt_info = DataAPI.OptGet(optID='', contractStatus=[u"DE", u"L"], field=fields_info, pandas="1")
    opt_info = opt_info.set_index(u"optID")
    
    cal = Calendar('China.SSE')
    dates = cal.bizDatesList(beginDate, endDate)
    histSkew = pd.DataFrame(0.0, index=map(Date.toDateTime, dates), columns=['SKEW30','SKEW60'])
    histSkew.index.name = 'tradeDate'
    for date in histSkew.index:
        try:
            skew30,skew60 =  calDaySkewIndex(Date.fromDateTime(date), opt_info)
        except:
            skew30,skew60 =  calDaySkewIndex(Date.fromDateTime(date), opt_info)
            histSkew = histSkew.drop(date)
            continue
        histSkew['SKEW30'][date] = skew30
        histSkew['SKEW60'][date] = skew60
        
    SKEW = pd.DataFrame()
    SKEW = histSkew[['SKEW30','SKEW60']]
    return SKEW

def plotSkewIndex():
    cal = Calendar('China.IB')

    start = Date(2015, 2, 9)
    end = Date.todaysDate()

    # Dates
    end = Date.todaysDate()
    end = cal.advanceDate(end,'-0B',BizDayConvention.Preceding)    
    #start = cal.advanceDate(end,'-60B',BizDayConvention.Preceding)   

    hist_skew = getHistDailySkewIndex(start, end)     # iVIX计算
    ax = hist_skew['SKEW30'].plot(figsize=(12,5), style='-')
    ax.set_ylabel('SKEW')
    # ax.set_ylim(10,65);
    return hist_skew.tail(10)
```
