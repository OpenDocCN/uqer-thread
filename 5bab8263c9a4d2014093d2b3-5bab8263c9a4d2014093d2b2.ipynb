    协整与相关性的不同在于相关性强调价格比回归，而协整强调价差回归，相关性强不等于协整关系成立。回到协整，首先需要基础时间序列要平稳，这个直接带公式，看P值，通过平稳性检验后，是否协整配对任然带公式，看P值。理论繁复，实际操作简单，可以回顾下Minghong大神原文，介绍非常清楚，上代码。https://uqer.datayes.com/v3/community/share/576a49f7228e5b819ba5607e原文链接，怎么加超链接，汗汗汗！

```python
import numpy as np
import pandas as pd
from pandas import *
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.stattools import coint
beginDate="20171202"
endDate="20180502"
universe = set_universe(IndSW.YinHangL1, '20160502')
print universe

def testStationarity(data):
    adftest = adfuller(data)
    result = pd.Series(adftest[0:4], index=['Test Statistic','p-value','Lags Used','Number of Observations Used'])
    for key,value in adftest[4].items():
        result['Critical Value (%s)'%key] = value
    return result

sec = []
#板块内循环，原始数据肯定不平稳了，直接一阶差分，通过P值过滤掉非平稳
for i in range(len(universe)):
    data1=DataAPI.MktEqudAdjGet(tradeDate=u"",secID=universe[i],ticker=u"",beginDate=beginDate,endDate=endDate,isOpen="",field=u"closePrice",pandas="1")
    diffx=data1.diff(1)
    diffx.dropna(inplace=True)
    diffx=np.array(diffx).T[0]
    zz = adfuller(diffx)
    if zz[1] < 0.005:
        sec.append(universe[i])
    
print sec
```
