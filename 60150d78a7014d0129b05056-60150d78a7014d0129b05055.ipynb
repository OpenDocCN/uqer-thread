# 抱团研究之策略 - 跟随基金重仓股

作为第二篇，本章节尝试利用基金重仓股信息，做简单跟随策略。有些步骤仍过于简单，希望大家多分享观点。

大体思路：
1） 利用公募基金季报中的持股披露信息，筛选基金重仓股
2） 在季报公布后的月底，分别按前10/30/50/100大持股，以等权、基金持股市值、自由流通市值的权重，进行季度调仓

结论：
利用基金公告信息中的持股信息可以提升策略收益


### 第一步 准备数据

```python
import datetime, math
import pandas as pd
import numpy as np
```

```python
# 计算持仓市值函数
def fundHoldingRatio_get(report_dt, trade_dt, df_fund_name, QtrReport_only = True):
    # 取所有基金的股票持仓信息
    df = DataAPI.FundHoldingsGet(reportDate=report_dt, secType="E",field=['secID','SecShortName','reportDate','holdingSecID','holdingTicker','marketValue'])

    # 如有仅用季报信息，则只可见前10大持股，但信息于季度后一个月可得
    if QtrReport_only == True:
        df['marketValueRank'] = df.groupby(['reportDate','secID'])['marketValue'].rank(ascending=False)
        df = df[df['marketValueRank']<=10] 

    # 仅保留A股
    df['exchange'] = df['holdingSecID'].apply(lambda x: x[-4:])
    df = df[df['exchange'].isin(['XSHE','XSHG'])] 

    # 利用分级信息剔除重复
    df_nodup = df.merge(df_fund_name, how='left', on='secID')
    df_nodup = df_nodup.drop_duplicates(subset=['holdingSecID','reportDate','secShortName_nodup','marketValue'])

    # 计算各股票的基金持股总市值
    df_holding_pos = df_nodup.groupby('holdingSecID')['marketValue'].sum().reset_index().rename(columns={'holdingSecID':'secID',
                                                                                                   'marketValue':'fundHolding'})
    # 取当季最后一个交易日的自由流动市值
    df_holding_pos = df_holding_pos.merge(DataAPI.MktEqudGet(secID=df_holding_pos['secID'],
                                                             tradeDate=trade_dt,field=['secID','negMarketValue','marketValue']),
                                          how='left', on='secID')
    # 基金持股比率=基金持股市值/自由流动市值
    df_holding_pos['fundHoldingRatio'] = df_holding_pos['fundHolding'] / df_holding_pos['negMarketValue']

    df_holding_pos = df_holding_pos.merge(DataAPI.EquGet(secID=df_holding_pos['secID'],
                                                         field=['secID','secShortName','ListSector','ListSectorCD']),
                                          how='left', on='secID')
    #df_holding_pos = df_holding_pos[df_holding_pos['ListSectorCD']!=4]

    # 仅保留基金持股比率>=1%的股票，减少输出数据
    df_holding_pos = df_holding_pos[df_holding_pos['fundHoldingRatio'] >= 0.01]
    df_holding_pos = df_holding_pos.sort_values(by='fundHolding',ascending=False)
    df_holding_pos['reportDate'] = report_dt
    return df_holding_pos[['reportDate','secID','fundHoldingRatio','fundHolding','negMarketValue','secShortName','ListSector']]

# 取季末交易日
trading_cal = DataAPI.TradeCalGet(exchangeCD=u"XSHG,XSHE",beginDate=u"20050101",endDate=datetime.datetime.today(),field=['calendarDate','isQuarterEnd','isOpen']).drop_duplicates()
trading_cal['MthID'] = trading_cal['calendarDate'].apply(lambda x: x[:7])
trading_cal_qtr = trading_cal[trading_cal['isQuarterEnd']==1]

# 取季报报告日期
ME_dates = trading_cal.sort_values(by='calendarDate').drop_duplicates(subset=['MthID'], keep='last')[['MthID','calendarDate']].rename(columns={'calendarDate':'ME_report_date'})
trading_cal_qtr = trading_cal_qtr.merge(ME_dates, how='left', on='MthID').rename(columns={'calendarDate':'ME_trade_date'}).set_index('MthID')

# 保存为数据字典
dict_report_trade_dt = dict(zip(trading_cal_qtr['ME_report_date'], trading_cal_qtr['ME_trade_date']))
trading_cal_qtr.tail()

# 取基金分级信息
df_fund_name = DataAPI.FundGet(field=['secID','secShortName','className']) 
df_fund_name['secShortName_nodup'] = df_fund_name.apply(lambda x: x['secShortName'] if pd.isnull(x['className']) else x['secShortName'][:-1-len(x['className'])], axis=1)

df_ratio_total = pd.DataFrame()
for qtrID in trading_cal_qtr.index:
    report_dt, trade_dt = trading_cal_qtr.loc[qtrID, 'ME_report_date'], trading_cal_qtr.loc[qtrID, 'ME_trade_date']
    #print('Pulling Fund Holding Ratios for quarter {:s} - report date {:s} - trading date {:s}'.format(qtrID, report_dt, trade_dt))
    df_ratio_total = pd.concat([df_ratio_total, fundHoldingRatio_get(report_dt=report_dt, trade_dt=trade_dt, df_fund_name = df_fund_name, QtrReport_only = True)])
    
output_file = 'Fund_Holding.csv'
df_ratio_total.to_csv(output_file, encoding='GBK', index=False)
print('Output to {:s}'.format(output_file))
```

### 第二步 策略日历

首先需要明确策略频率，考虑到季报信息披露时效性强且间隔平均，本文尝试最简单的季度跟随策略，即在季报公布的当月月底，利用基金持仓市值排序构建组合。

策略日历中，有如下四个日期，以2019年1季报信息为例：
- QE_tradeDate - 季末最后交易日 （2019-03-29）
- report_asOfDate - 季报数据截止日期 （2019-03-31）
- report_publishDate - 季报最晚公布日期，新一季度的十五个工作日内 （2019-04-22）
- rebalDate - 再平衡日期，季报公布当月的最后一个交易日 （2019-04-30）

```python
# 取基础交易日历
trading_cal = DataAPI.TradeCalGet(exchangeCD=u"XSHG,XSHE",beginDate=u"20050101",endDate=datetime.datetime.today(),field=['calendarDate','isQuarterEnd','isOpen']).drop_duplicates()

# 取季末交易日
trading_cal['MthID'] = trading_cal['calendarDate'].apply(lambda x: x[:7])
trading_cal['QtrID'] = trading_cal['calendarDate'].apply(lambda x: '{:s}Q{:n}'.format(x[:4],math.ceil(int(x[5:7])*1.0/3)))
trading_cal_qtr = trading_cal[trading_cal['isQuarterEnd']==1]

# 取季报报告日期
ME_dates = trading_cal.sort_values(by='calendarDate').drop_duplicates(subset=['MthID'], keep='last')[['MthID','calendarDate']].rename(columns={'calendarDate':'report_asOfDate'})
trading_cal_qtr = trading_cal_qtr.merge(ME_dates, how='left', on='MthID').rename(columns={'calendarDate':'QE_tradeDate'})

# 取季报最晚公布日期（季度后十五个工作日内）
trading_cal['tday_from_QE'] = trading_cal.groupby('QtrID')['isOpen'].cumsum()
qtr_report_dt = trading_cal[(trading_cal['tday_from_QE']==15)&(trading_cal['isOpen']==1)].rename(columns={'calendarDate':'report_publishDate','QtrID':'curr_QtrID','MthID':'report_publishMthID'})
qtr_report_dt['report_QtrID'] = qtr_report_dt['curr_QtrID'].shift(1)
qtr_report_dt = qtr_report_dt.set_index('report_QtrID')[['report_publishDate','report_publishMthID']]
trading_cal_qtr = trading_cal_qtr.merge(qtr_report_dt, how='left', left_on='QtrID', right_index=True)

# 取策略再平衡日期（季报公布日后的月末交易日）
ME_trading_dates = trading_cal[trading_cal['isOpen']==1].sort_values(by='calendarDate').drop_duplicates(subset=['MthID'], keep='last').set_index('MthID')[['calendarDate']].rename(columns={'calendarDate':'rebalDate'})
trading_cal_qtr = trading_cal_qtr.merge(ME_trading_dates, how='left', left_on='report_publishMthID', right_index=True)
trading_cal_qtr = trading_cal_qtr.set_index('QtrID')[['QE_tradeDate','report_asOfDate','report_publishDate','rebalDate']]

# 保存为数据字典
dict_report_trade_dt = dict(zip(trading_cal_qtr['report_asOfDate'], trading_cal_qtr['QE_tradeDate']))

trading_cal_qtr.tail(10)
```

### 说明
继续以2019年1季报信息为例，具体投资流程为：
- 2019-04-22，汇总所有基金1季报，计算基金持仓市值
- 2019-04-30，调整策略组合内个股头寸，持有3个月
- 2019-07-19，汇总所有基金2季报，计算基金持仓市值
- 2019-07-31，调整策略组合内个股头寸，继续持有新组合3个月

所以由2019年1季报中的信息构建的组合，在2019-04-30至2019-07-31之间被持有。
根据这样的规律，需要进一步计算各季报信息对应的实际收益率。

### 第三步 季报信息对应的实际收益率
计算实际收益率时需要考虑信息真实可获得时间与投资之间的间隔，例如一季报中3月底的信息，对应4月-7月内持有股票的收益率

```python
# 填充交易日期以方便合并信息
df_ratio_total['tradeDate'] = df_ratio_total['reportDate'].apply(lambda x: dict_report_trade_dt[x])
df_ratio_total.tail()
```

```python
# 计算未来实际收益率
def fwd_ret_calc(begin_date = '20091231', end_date = '20210131', secID_list = None, date_list = None):
    # 如果有secID列表，可以缩小范围 
    if secID_list is None:
        df_p = DataAPI.MktEqumAdjGet(secID='', beginDate=begin_date, endDate=end_date, field=['secID','endDate','closePrice'])
    else:
        df_p = DataAPI.MktEqumAdjGet(secID=secID_list, beginDate=begin_date, endDate=end_date, field=['secID','endDate','closePrice'])
    
    # 未来收益率 - 例如3月底的信息，对应4月-7月内持有股票的收益率
    for fwd_n in [1,2,3,4]:
        df_p.loc[(df_p['secID'] == df_p['secID'].shift(-fwd_n)), 'closePrice_fwd'+str(fwd_n)] = df_p['closePrice'].shift(-fwd_n) 
    df_p['fwdret_3m'] = df_p['closePrice_fwd4'] / df_p['closePrice_fwd1'] - 1
    df_p['fwdret_2m'] = df_p['closePrice_fwd3'] / df_p['closePrice_fwd1'] - 1
    df_p['fwdret_1m'] = df_p['closePrice_fwd2'] / df_p['closePrice_fwd1'] - 1
    
    # 如果有交易日列表，可以缩小输出数据
    if date_list is not None:
        df_p = df_p[df_p['endDate'].isin(date_list)]
    
    return df_p[['secID','endDate','fwdret_3m','fwdret_2m','fwdret_1m']]
df_p = fwd_ret_calc(date_list = df_ratio_total['tradeDate'].unique())
df_p.tail()
```

```python
# 将实际的未来收益率与基金重仓股信息汇总
df_total = df_p.merge(df_ratio_total[['secID','secShortName','tradeDate','fundHoldingRank','fundHolding','negMarketValue']].rename(columns={'tradeDate':'endDate'}),how='left', on=['secID','endDate'])

# 分别计算出各期季报中对应的前10/30/50/100大持股
for top_num in [10, 30, 50, 100]:
    df_total['Top{:n}'.format(top_num)] = df_total['fundHoldingRank'] <= top_num

# 检查是否有缺少实际未来收益率的情况
df_count = df_total[pd.notnull(df_total['fwdret_3m'])].groupby('endDate')[['Top10','Top30','Top50', 'Top100']].sum()
df_count[df_count['Top100']<100]
```

所以下面取这些股票的名称。

```python
df_missing_total = pd.DataFrame()
for dt in df_count[df_count['Top100']<100].index:
    secIDs_missing_fwdret = df_total[(df_total['endDate']==dt) & (df_total['Top100']) & pd.isnull(df_total['fwdret_3m'])]['secID']
    df_missing = df_total[(df_total['endDate']==dt) & (df_total['fundHoldingRank'] <=100 ) & (df_total['secID'].isin(secIDs_missing_fwdret))] 
    df_missing_total = pd.concat([df_missing_total, df_missing])
df_missing_total
```

- 2013Q2后仍持有美的电器至退市，收益率计算区间为2013年7月底至2013/09/18退市，同理适用于2015Q3的招商地产

```python
# 美的电器
adj_bool = (df_total['endDate']=='2013-06-28') & (df_total['secID']=='000527.XSHE')
df_total.loc[adj_bool,'fwdret_3m'] = df_total.loc[adj_bool,'fwdret_2m']
df_total.loc[adj_bool,'adj_flag'] = 1

# 宏源证券
adj_bool = (df_total['endDate']=='2014-12-31') & (df_total['secID']=='000562.XSHE')
df_total.loc[adj_bool,'fwdret_3m'] = 0.0
df_total.loc[adj_bool,'adj_flag'] = 1

# 招商地产
adj_bool = (df_total['endDate']=='2015-09-30') & (df_total['secID']=='000024.XSHE')
df_total.loc[adj_bool,'fwdret_3m'] = df_total.loc[adj_bool,'fwdret_2m']
df_total.loc[adj_bool,'adj_flag'] = 1

# 大华农
adj_bool = (df_total['endDate']=='2015-09-30') & (df_total['secID']=='300186.XSHE')
df_total.loc[adj_bool,'fwdret_3m'] = 0.0
df_total.loc[adj_bool,'adj_flag'] = 1

# 检查是否修改正确
df_total[df_total['adj_flag']==1]
```

### 第四步 策略收益
每个季度，根据季报中基金持股总市值选出股票池，以不同权重构建组合，持有三个月，直至下次基金季报信息公布月。

```python
# 股票池重个股数量 - top_num = 100，即每次选取前100大基金持股
top_num = 100

# 检查数据完备 - 各季报期中，所选出的个股是否都有实际收益率数据
df_pool = df_total[df_total['Top{:n}'.format(top_num)]].dropna(subset=['fwdret_3m'])
print('前{:n}大基金持股实际收益率信息缺失的报告期个数 : {:n}'.format(top_num, (df_pool.groupby(['endDate'])['fundHolding'].count()<top_num).sum()))
```

```python
# 设置权重：weight_fundHolding（按基金持股市值），weight_mktcap（按自由流通市值），weight_equal（等权）
df_pool['weight_fundHolding'] = df_pool.groupby(['endDate'])['fundHolding'].transform(lambda x: x / x.sum())
df_pool['weight_mktcap'] = df_pool.groupby(['endDate'])['negMarketValue'].transform(lambda x: x / x.sum())
df_pool['weight_equal'] = df_pool.groupby(['endDate'])['negMarketValue'].transform(lambda x: 1.0 / len(x))
```

```python
# 计算策略未来实际收益率 - 股票池中个股权重 * 个股未来实际收益率
df_strategy_fwdret = pd.DataFrame(index=trading_cal_qtr['QE_tradeDate'])
for weight_method in ['weight_fundHolding','weight_mktcap','weight_equal']:
    ret = df_pool.groupby(['endDate']).apply(lambda x: (x[weight_method] * x['fwdret_3m']).sum())
    df_strategy_fwdret[weight_method] = ret
df_strategy_fwdret = df_strategy_fwdret.loc[df_pool['endDate'].min():]

df_strategy_fwdret = df_strategy_fwdret.merge(trading_cal_qtr[['QE_tradeDate','rebalDate']], how='left', left_index=True, right_on='QE_tradeDate').set_index('rebalDate')
df_strategy_fwdret.tail()
```

```python
# 计算策略当前累计收益率 - 注意1个季度的时滞
df_strategy_cumret = np.cumprod(1 + df_strategy_fwdret[['weight_fundHolding','weight_mktcap','weight_equal']].shift(1))

# 为策略初期填充净值起点
first_rebal_date = df_strategy_fwdret.index[0]
df_strategy_cumret.loc[first_rebal_date,:] = np.array([1.0,1.0,1.0])

df_strategy_cumret.tail()
```

### 第五步 对比指数
选取三个指数进行对比：上证综指、上证50、沪深300

```python
# 指数信息
idx_ticker = ['000001','000016','000300']
DataAPI.IdxGet(ticker=idx_ticker, field=['secID','ticker','secShortName'])
```

```python
# 策略同期的指数累计收益率
df_p_idx = DataAPI.MktIdxmGet(beginDate=first_rebal_date,ticker=idx_ticker,field=['indexID','ticker','secShortName','endDate','closePrice'])
df_p_idx = df_p_idx.pivot_table(values='closePrice', index='endDate', columns='ticker')
df_p_idx.columns = ['Idx_' + col for col in df_p_idx.columns]
df_p_idx = df_p_idx / df_p_idx.iloc[0]
df_p_idx.tail()
```

```python
# 数据合并展示
df_strategy_summary = df_strategy_cumret.merge(df_p_idx, how='left', left_index=True, right_index=True)
df_strategy_summary.plot(figsize=(12,8));
df_strategy_summary.tail()
```
