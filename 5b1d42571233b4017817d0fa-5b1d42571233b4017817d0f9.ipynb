# 第6课 组合优化

$$minimize:c$$
$$subject \ to:Ax=b$$
$$----------------------------------------------------$$
一个简单的组合优化问题:最大化组合的预期收益率(这里用信号值代替),组合当中,个股权重下限为0,上限为0.05,全额投资.那么数学表达为:
$$minimize:-w'f$$
$$subject \ to:\ -w_{i}<=0$$
$$w_{i}<=0.05$$ 
$$全额投资sum_{i}w_{i} = 1即：$$
$$-sum_{i}w_{i}<=-1$$
$$sum_{i}w_{i}<=1$$

```python
import pandas as pd
import numpy as np
from cvxopt import matrix, solvers

# 1. 设置股票池
benchmark = 'HS300'
date = '20171229'
universe = set_universe(benchmark, date)#获得指定日期的指定成分股，相当于前面使用API获取指定日期的成分股

# 2. 获取信号(以PE为例)
factor = DataAPI.MktStockFactorsOneDayGet(tradeDate=date, secID=universe, field='secID,PE').set_index('secID').dropna()

alpha = factor['PE']
#每只个股最大权重为0.05，全额投资，也即选出'PE'最大的20只个股进行投资
print alpha.sort_values().tail(20).sum()*0.05
```

```python
#组合优化的方式来解决：
#分别求出c、A、b输入矩阵，代入公式得到最优解

c = -1 * matrix(alpha)#要求alpha的最大值，转化为求-alpha的最小值

min_weight = pd.Series(0, index=alpha.index)
max_weight = pd.Series(0.05, index=alpha.index)
N = len(alpha)

# a1 个股下限约束
a1 = np.mat(-1 * np.identity(N))#全为-1的对角阵
#np.mat(a)创建矩阵
#np.identity(N),获取方阵，也即标准意义的单位阵
b = (-1 * min_weight).tolist()

# a2 个股上限约束
a2 = np.mat(np.identity(N))#全为1的对角阵
b = np.append(b, max_weight.tolist())

# a3 权重之和下限
a3 = np.mat(-1 * np.ones(N))
b = np.append(b, -1)

# a4 权重之和上限
a4 = np.mat(np.ones(N))
b = np.append(b, 1)

A = matrix(np.vstack((a1, a2, a3, a4)))#纵向相加
b = matrix(b, tc='d')

sol = solvers.lp(c, A, b)
print(sol['primal objective'])
#计算出来的值与前面一样
```

cvxopt例子1:
$$minimize: 2x_1+ x_2$$
$$subject \ to\ -1 * x_{1} + 1 * x_{2}<=1,$$
$$-1 * x_{1} - 1 * x_{2} <=-2$$
$$0 * x_{1} + 1 * x_{2}<=0$$
$$1 * x_{1} - 2 * x_{2} <= 4$$

```python
import pandas as pd
from cvxopt import matrix, solvers
#cvxopt,Convex Optimization（凸规划）模块
#求最小值
#约束条件都要转成小于等于形式？
c = matrix([2.0, 1.0 ])#目标矩阵
A = matrix([[-1.0, -1.0, 0.0, 1.0], [1.0, -1.0, -1.0, -2.0]])#x1与x2的系数矩阵
b = matrix([1.0, -2.0, 0.0, 4.0 ])
sol = solvers.lp(c, A, b)
print(sol['x'])#获得最优解矩阵
print(np.dot(sol['x'].T, c))#最优解矩阵的转置与目标矩阵的点乘
# dot函数是np中的矩阵乘法
# x.dot(y) 等价于 np.dot(x,y)
# x是m*n 矩阵 ，y是n*m矩阵
# 则x.dot(y) 得到m*m矩阵
print(sol['primal objective'])
```

cvxopt例子2:
$$z = x_1+ 2x_2的取值范围$$
$$x_{1} <=2$$
$$x_{2} <=2$$
$$x_{1} + x_{2} >=2$$


```python
import numpy as np
from cvxopt import matrix, solvers

A = matrix([[1.0, 0.0, -1.0], [0.0, 1.0, -1.0]])
b = matrix([2.0, 2.0, -2.0])
c = matrix([1.0, 2.0])
d = matrix([-1.0, -2.0])#求最大值时使用，转化成求最小值

sol1 = solvers.lp(c,A,b)
print(sol1['x'])
min = np.dot(sol1['x'].T, c)

sol2 = solvers.lp(d,A,b)
print(sol2['x'])
max = -np.dot(sol2['x'].T, d)

print('min=%s,max=%s'%(min[0][0], max[0][0]))
```

二次规划问题:
$$minimize: 0.5x.TPx+q.Tx$$
$$subject\ to \ : G*x<=h$$
$$Ax=b$$
$$其中.T表示转置，P,q,G,h,A,b为输入矩阵，该问题求解采用QP算法$$

二次型规划问题,cvxopt例子3:
$$minimize: 2x_{1}*x_{1}+ x_{2}*x_{2}+x_{1}*x_{2}+x_{1}+x_{2}$$
$$subject\ to \ : x_{1} >=0$$
$$x_{2} >=0$$
$$x_{1} +  x_{2} = 1$$


```python
from cvxopt import matrix, solvers

Q = 2*matrix([[2, .5], [.5, 1]])#
p = matrix([1.0, 1.0])#
G = matrix([[-1.0,0.0],[0.0,-1.0]])#x1与x2的系数矩阵
h = matrix([0.0,0.0])
A = matrix([1.0, 1.0], (1,2))
b = matrix(1.0)

sol=solvers.qp(Q, p, G, h, A, b)
print(sol['x'])
print(sol['primal objective'])
```

二次型规划问题,cvxopt例子4:
$$minimize: 0.5x_{1}*x_{1}+3x_{1}+4x_{2}$$
$$subject\ to \ : x_{1},x_{2} >=0$$
$$x_{1}+ 3x_{2} >=15$$
$$2x_{1}+ 5x_{2} <=100$$
$$3x_{1} +  4x_{2} <= 80$$


```python
from cvxopt import matrix, solvers

P = matrix([[1.0, 0.0], [0.0, 0.0]])
q = matrix([3.0, 4.0])
G = matrix([[-1.0, 0.0, -1.0, 2.0, 3.0], [0.0, -1.0, -3.0, 5.0, 4.0]])
h = matrix([0.0, 0.0, -15.0, 100.0, 80.0])

sol=solvers.qp(P, q, G, h)
print(sol['x'])
print(sol['primal objective'])
```
