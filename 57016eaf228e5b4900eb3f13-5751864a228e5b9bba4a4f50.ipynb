# 线性回归建模中残差异方差性的分析和处理

## 1 问题背景

线性回归模型广泛应用于经济和金融的量化分析中。本文主要基于Coursera平台Data Science专题的线性回归模型课程的材料，加上本人的的学习和实践心得，对残差异方差性的各种情况和处理方法进行讨论。

线性回归方程的通式如下：

$$ Y = \beta_1*X + \beta_2 + \epsilon  $$ 

其中Y为因变量，X为自变量，$\beta_1$ 为自变量的系数，$\beta_2$为截距，$\epsilon$为残差项。 

在建模过程中，我们得到一系列数据点的X和Y值，对参数$\beta_1$及$\beta_2$进行估计。当应用线性回归对数据进行建模的时候，我们实际上假设了因变量Y的取值由线性部分( $\beta_1*X +\beta_2$ )和随机部分（服从正态分布的$\epsilon$）决定。对于残差项的分析，是分析模型合理性的重要指标。在线性回归模型中，残差应满足白噪声假设（White Noise Condition）:

（1） 残差独立同分布（independent and identical distribution，iid），且无自相关性；

（2） 残差和自变量X不相关;

（3） 残差的均值为0，方差为常数。

在统计学中，白噪声随机序列是指一组无自相关性，且有相同分布的随机序列。理论上，白噪声假设不要求随机变量服从正态分布，而可以是任意分布。但基于中心极限定理，假设残差服从正态分布是一个合理的近似。

基于以上白噪声假设的第3条，当残差方差为常数时，我们称残差具有同方差性（homoscedasticity）；当残差方差不是常数时，称残差具有异方差性（heteroscedasticity）。

异方差性的存在意味着违反了线性回归模型的白噪声假设。因此，对于异方差性的分析有助于我们理解数据的问题或特征，而对于异方差性的修正则有助于提高模型参数估计的准确度。


## 2. 数据可视化，离群值和残差异方差性的判断

在进行线性回归建模前，一般要先通过散点图来观察数据的基本特征。著名的安斯库姆四重奏（Anscombe's quartet）展示了在线性回归模型中具有相同的统计特征，但数据分布明显不同的四个例子，用于说明线性回归建模前进行数据可视化分析的重要性：

![](http://2.bp.blogspot.com/_IFzDPHUxHI0/SG0ocfCh01I/AAAAAAAAADI/VAqSLJd0dLc/s400/anscombe_quartet.gif)

一般在进行可视化分析的时候，我们除了关注数据是否存在明显的线性相关特征外，还需要观察离群值的数量。离群值和残差异方差性是紧密相关的概念。通常，如果一个数据点为离群值，同时也意味着它对应的残差具有较大的方差，因此数据中的离群值数量较多的话，残差一般也会出现明显的异方差性。

关于线性回归的离群值的判断，有两个要点：

数据中存在少量的离群值是合理的。例如，当我们产生1000个服从标准正态分布的随机数，以距离均值大于两个标准差作为离群值判断标准，因为数据落在两个标准差之外的概率约为4.5%，意味这1000个抽样中大约会有45个离群值。此时如果我们去除这45个离群值来估计分布的方差，将会得到小于1的结论。因此，在删去离群值前应慎重考虑，除了因为存在少量离群值是合理的以外，离群值可能包含抽样或者数据的特征或者存在的问题。因此，如果数据中存在相当数量的离群值，应分析其成因，而非简单将其删去。

线性回归离群值（regression outlier）是指对线性回归模型参数估计有强影响力的离群值（influential outlier）。只有当一个离群值具有高杠杆值（high leverage）且有明显的偏差（significant discrepancy）时，它才有可能是具有强影响力的。对于一元回归而言，只有当数据点出现在图的右下方时，它才有可能是有强影响力的。

对于多元回归模型，不能通过简单可视化来判断离群值的数量。可以通过cook’s distance或者已添加变量图（added variable plot）来进行判断。

在完成线性回归建模后，我们可以通过残差图（residual plot）来判断残差的异方差性：

![](http://i1.wp.com/condor.depaul.edu/sjost/it223/documents/resid-plots.gif)

如果残差在图X轴两侧均匀分布，则意味着异方差性较弱；而如果呈现出明显的不均匀分布，则意味着存在明显的异方差性。


## 3. 残差异方差性的成因分析

根据本人学习和经验，异方差性的成因总结为以下四种：

（1） Y和X本身存在非线性关系；
（2） Y本身存在显著的自相关性；
（3） 残差中包含和因变量X线性相关，但未被模型考虑的变量，导致残差和X相关；
（4） 受数据采样精度或流程所限，数据抽样时不独立，或环境发生显著变化，残差不满足独立同分布的假设。


### 3.1 Y和X本身存在非线性关系引起的残差异方差性

对于第1种情况，例如Y与X本身满足的是如下函数关系：Y=Sin(X)

因为Y和X存在显著地非线性相关，因此此时用简单线性回归建模，将会出现欠拟合（underfitting）的问题。欠拟合可理解为在建模过程中，选择的模型过于简单，因而无法反映数据的精细结构。在大部分实际问题分析中，我们不可能准确知道Y和X的关系。一种处理的思路是，我们可以在模型中考虑高次多项式来拟合Y和X非线性相关部分（多项式回归）

$$ Y = \beta_1*X + \beta_2*X^2 + \beta_2*X^3 + … +\beta_n*X^n  $$

用高次多项式作为基组进行拟合，等价于对Y进行泰勒展开，理论上项数考虑越多，对于Y的拟合精度越高。而使用高次多项式的主要问题在于可能出现过拟合（overfitting），即估计模型参数时，把相当多噪声考虑其中，从而影响模型的泛化能力。在实际中，一般只能使用二次多项式来对非线性部分进行拟合。在下文讨论稳健回归（robust regression）时将看到具体例子。

在《机器学习实战》一书中线性回归模型部分，还提到另外一种拟合非线性部分的回归模型，称为局部加权线性回归 (Locally Weighted Linear Regression，LWLR)。该模型通过在最小二乘法得到的参数估计表达式中，加入一个核函数矩阵，从而在进行参数估计时，给予附近的点一定的权重来拟合数据的非线性相关部分：

$$ \beta = \dfrac{X^T}{X^TW X} Y $$

其中W通常选择为高斯核函数矩阵，该矩阵为对角矩阵，对角线性的元素为：

$$ W (i, i)=exp{\left( -\dfrac{(x_{i}- x)^2}{2k^2}  \right)  } $$


其中$x_{i}$为数据点x附近的数据点，k决定了附近数据点的权重。拟合效果如下图所示：

![](https://upload.wikimedia.org/wikipedia/commons/thumb/d/d2/Loess_curve.svg/220px-Loess_curve.svg.png)

局部加权线性回归模型中，可以通过权衡偏差和系数估计方差来判断k的合理值。偏差（bias）为预测值和真实值之间的误差，而系数估计方差（variance）则是利用不同的训练集得到一组参数估计值后，用这组估计值计算得到的方差。通常方差的估计通过交叉验证 (cross-validation)来实现。欠拟合的模型一般具有高偏差和低方差，而过拟合的模型则具有低偏差和高方差，合理的模型应该同时具有较低的偏差和方差，如下图所示：

![](http://img.my.csdn.net/uploads/201107/28/0_1311833568hD6X.gif)

### 3.2 因变量Y本身存在显著自相关性引起的残差异方差性

在经济学中，因变量Y本身存在显著自相关性最著名的例子是股票的波动率的聚合（volatility cluster），如下图所示，自变量X为时间，因变量Y为波动率。图中显示当波动率上升时，下一时刻波动率也倾向于上升，呈现明显的自相关性。

![](http://l3.yimg.com/bt/api/res/1.2/7OHacGZxCO_hGlCYxLt3rw--/YXBwaWQ9eW5ld3NfbGVnbztxPTg1/https://media.dailyfx.com/illustrations/2014/05/07/forex-why-is-volatility-so-low-and-how-do-we-trade-high-volatility_body_Picture_5.png)

对于因变量本身存在显著自相关性，且自变量为时间的数据，我们应该采用时间序列分析建模。对于波动率，我们有EWMA，ARCH和GARCH模型等进行建模和分析。在这里不对这些模型的细节进行探讨。

更一般地，因变量自相关性的存在意味着因变量Y和残差项均不满足正态分布的假设。一种解决思路是利用Box-Cox transformation，通过数据变换使得因变量更接近正态分布。此外，下面介绍的广义最小二乘法也有助于消除残差的自相关性。但需要注意的是，这些处理技巧能够降低因变量/残差存在自相关性的影响，但如果数据本身存在很强的异方差性，那么应仔细分析数据的抽样流程和统计特征，再判断线性回归是否适用。


### 3.3 残差和自变量X线性相关

残差和因变量X相关，意味着残差中存在未被模型考虑的变量，且该变量和X相关。而残差和因变量X相关引起的异方差性，对于模型预测能力的影响，取决于被忽略的变量是否和因变量Y线性相关。

我们可以通过一个简单的例子来理解这个问题。假设我们以雪糕的销量作为因变量Y，季节作为自变量X进行线性回归建模。基于生活常识，夏天应该和雪糕的销量存在明显的正相关，而冬天应该和雪糕销量存在明显的负相关。

除了季节以外，还会有很多变量影响雪糕的销量，例如温度。温度显然和季节相关，又很有可能和雪糕销量相关。在统计学中，一般称这样的变量为混淆变量（confounding variable）。混淆变量是指和其它变量同时相关，导致我们无法准确其它变量的关系的变量。

在建模中，如果存在混淆变量，即和Y及X都显著相关的变量进入残差中，将会导致忽略变量偏差（omiited variable bias），造成系数和模型的估计有偏，模型的预测能力将会被大大影响。

混淆变量进入残差项中，通常是因为建模时对自变量共线性（colinearity）处理不当。自变量共线性是指自变量之间存在线性相关，如果存在多个自变量线性相关，则称为多重共线性（multicolinearity）。自变量共线性不会影响模型的整体预测能力，但会影响对于单个变量的系数估计。当共线性较为显著时，自变量的协方差矩阵是一个奇异矩阵，影响最小二乘法对于参数的估计。

在处理共线性问题时，一个重要的原则是不能忽略和Y显著相关的变量（可以通过该变量加入模型与否会明显改善参数显著性检验来判断）。只有在该变量和Y无明显相关性时，我们才能考虑去掉这个变量。

在不去掉任何自变量的前提下，仍然有很多办法处理共线性问题，例如主成分分析，或者进行正则化处理的岭回归和Lasso回归等。自变量共线性处理是线性回归的另一个重要问题，在这里不再作深入讨论。

另外，在实际问题建模中，我们不可能知道所有的混淆变量，甚至可能存在无法测量混淆变量的情况。除了采用合适的建模方法来处理这些问题，更为根本的方法是提高抽样的数量和质量，来减少混淆变量的影响。


### 3.4 抽样时不独立或环境发生显著变化引起的残差异方差性

在对数据建模时，了解数据是如何采集的，有助于我们理解数据的特征和可能存在的问题，同时也可判断数据是否满足建模方法的假设，从而采用合适的建模手段。

在特殊情况下，当抽样流程存在严重问题，或者受抽样手段的限制，数据中的噪音较大，无法有效提取数据的特征时，那么线性回归模型将难以对数据进行解释和预测。


## 4 残差异方差性的处理

在参数估计中，我们一般通过最小二乘法，以最小化残差的方差为目标进行参数估计：

$$ min{F} = min{{(Y-\beta*X)^T(Y-\beta*X)} }  $$

求上式最小值等价于令函数F对$\beta^T$的导数为0：

$$ \dfrac{\delta F}{\delta\beta^T} = X^T(Y-\beta*X)=0  $$

因此得到系数\beta的表达式

$$ \beta = \dfrac{X^T}{X^T X} Y  $$

从上述表达式可以看出，系数$\beta$为Y的线性组合，因为Y满足正态分布（线性部分+正态分布残差），因此系数满足联合正态分布，这是一般系数显著性检验的基础。

因为最小二乘法估计的目标是最小化残差的方差，因此当残差存在异方差性时，系数的估计值将受那些具有较大方差的数据点影响，从而具有较大的方差，影响后续的模型的显著性检验和离群值判断。为了降低残差异方差性对参数估计的影响，以下介绍稳健回归（robust regression）和广义最小二乘法(generalized least square)两种方法。

### 4.1 稳健回归

为了降低具有较大的方差的数据点对参数估计的影响，稳健回归选择最小化以下的目标函数：

$$ min{S} = min{{|Y-F(X)|} }  $$

一种方式是选择F(X) = aX^2 + bX + c，用二次多项式来拟合Y和X的非线性相关部分。

稳健回归的优点在于它放弃了最小二乘法的平方和形式的目标函数，从而降低了残差方差较大的数据点对参数估计的影响，且能够拟合Y和X的非线性相关部分。

稳健回归的缺点在于参数a，b和c可能存在多组解能够最小化上述目标函数，从而造成参数估计不稳定；此外，使用二次多项式也导致了过拟合的风险；而稳健最重要的问题是，最小二乘法得到的参数估计表达式满足联合正态分布，这是一般进行线性回归建模系数显著性检验的基础假设。但稳健回归无法保证参数符合联合正态分布，因此一般参数显著性检验的方法未必适用。

### 4.2 广义最小二乘法

广义最小二乘法对最小二乘法的目标函数进行以下调整：

$$ min{f} = min{{(Y-\beta*X)^T\Omega^{-1}(Y-\beta*X)} }  $$

其中$\Omega^{-1}$为残差协方差矩阵的逆矩阵。

引入$\Omega^{-1}$的意义可作如下分析。因为协方差矩阵为半正定矩阵，因此可对其进行特征分解：

$$  \Omega=U^T\Sigma U  $$

其中$U$为正交矩阵，$U^T$为其转置，$\Sigma$为对角矩阵。

由此可得到$\Omega^{-1}$的特征分解：

$$  \Omega^{-1}=U^T\Sigma^{-1} U  $$

广义最小二乘法的表达式可写为：

$$ min{f} = min{{(Y-\beta*X)^TU^T\Sigma^{-1}U (Y-\beta*X)} }  $$

由此可看到，引入$\Omega^{-1}$的意义在于通过正交矩阵U把残差投影到一个新的线性空间，从而剔除其自相关性；同时$\Sigma^{-1}$可理解为赋予残差不同的权重（权重和残差的方差成反比），从而降低了残差异方差性的影响。

以上，和稳健回归相比，广义最小二乘法的优点在于最小二乘法目标函数的形式稍作调整，在降低残差异方差性影响的同时，也使得一般的系数显著性检验能够继续使用。广义最小二乘法的主要缺点在于其运算量较大，在建模过程中必须先用最小二乘法建模，得到残差的估计值，计算其经验协方差矩阵，再用于广义最小二乘法建模。



## 5. 残差异方差性分析和处理总结

以上我们对于残差异方差性的判断，成因，处理方法进行了探讨。需要强调的是，虽然以上介绍的稳健回归和广义最小二乘法能够降低残差异方差性对于参数估计的影响，但不经考虑使用这些方法，可能会掩盖了数据本身的特征（存在显著的非线性相关，因变量存在自相关性等），或者建模及抽样时出现的问题（混淆变量进入残差中，抽样不合理），从而使得线性模型未能准确提取数据特征，也不具备准确预测的能力。

因此，总结来说，只有在建模时对于数据的采集方式，数据的特征和各种建模方式的深入了解，才能保证我们对于残差异方差性有一个准确的认识，得到的是一个合理的线性回归模型。
