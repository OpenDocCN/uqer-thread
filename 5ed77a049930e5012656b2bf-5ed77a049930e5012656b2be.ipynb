```python
# 取沪深300成分股
hs300_frame = DataAPI.IdxConsGet(secID=u"",ticker=u"000300",isNew=u"",intoDate=u"",field=u"",pandas="1")
hs300_frame=hs300_frame[hs300_frame.isNew==1]
hs300_frame.head()
```

```python
import pandas as pd
# list格式的300个股票
tickers = list(hs300_frame.consTickerSymbol.values)

# 取该300个股票2017-05-01到2017-06-01的因子值
# 先取交易日历
calendar_frame = DataAPI.TradeCalGet(exchangeCD=u"XSHG,XSHE",beginDate=u"20170501",endDate=u"20170601",field=u"",pandas="1")
# 只取有交易的日期
calendar_frame = calendar_frame[calendar_frame.isOpen==1]
calendar_frame.head()
factor_frame_list = []
# 遍历每一天，拿到因子数据
for tdate in calendar_frame.calendarDate.values:
    factor_frame = DataAPI.MktStockFactorsOneDayGet(tradeDate=tdate,ticker=tickers,field=["ticker",'tradeDate','ROE', 'DAVOL10'],pandas="1")
    factor_frame_list.append(factor_frame)
# 合并所有天的因子值
tfactor_frame = pd.concat(factor_frame_list)
tfactor_frame.head()
```

```python
import numpy as np
import pandas as pd
tfactor_frame1 = tfactor_frame
tfactor_frame1.loc[0,'ROE'] = np.nan
tfactor_frame1.head()
```

```python
sw_map_frame = DataAPI.EquIndustryGet(secID=u"",ticker=tickers,industryVersionCD=u"010303",\
                            industry=u"",industryID=u"",field=u"ticker,secShortName,industryName1,isNew",pandas="1")
sw_map_frame = sw_map_frame[sw_map_frame['isNew']==1]
sw_map_frame.head()
```

```python
tfactor_frame1 = tfactor_frame1.merge(sw_map_frame,on='ticker',how='left')
tfactor_frame1['tradeDate']=pd.to_datetime(tfactor_frame1['tradeDate'])
tfactor_frame1
```

```python
tfactor_frame2=tfactor_frame1.groupby(['tradeDate','industryName1'])[['ROE','DAVOL10']].mean()
tfactor_frame2 = tfactor_frame2.rename(columns = {'ROE':'mean_ROE','DAVOL10':'mean_DAVOL10'})
tfactor_frame2['tradeDate']=pd.to_datetime(tfactor_frame1['tradeDate'])
tfactor_frame2
```

```python
tfactor_frame1 = tfactor_frame1.merge(tfactor_frame2,on=['tradeDate','industryName1'],how = 'left')
tfactor_frame1
```
