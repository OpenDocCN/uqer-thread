# 一致性风险度量之CVaR与ES

# 一.前言
- ## 关于风险的度量一直是我们投资活动的重点，从经典的Markowitz的方差，到现代流行的风险度量VaR,金融风险管理的方法随着金融市场的不断发展而日新月异
- ## 关于VaR的详细内容就不在这里赘述了，详情可以参考社区大神的帖子[在险价值(Value at Risk)风险度量下的资产组合前沿](https://uqer.datayes.com/community/share/57aa83f8228e5b9b95a88894)
- ## 简单回顾一下，VaR表示了“在正常的市场条件下，给定的置信区间的一个持有期内的最坏的预期损失”，衡量了市场正常波动情况下，投资组合可能的损失。
- ## VaR以最简单的形式将不同市场因子，不同市场的风险集成为一个数，较好的综合反映了潜在损失，其计算方法有很多，对风险因素的不同预测模型计算出来的风险数据各不相同。当然，它也有许多局限性，比如证券组合的风险不一定小于各证券风险之组合，市场极端情况下的损失没有衡量等.因此，有人提出了更多更完善的风险度量方法
- ## 本帖主要介绍了一种经典的`一致性风险度量方法`————`CVaR,即ES`，并用了一种简单的方法来`计算`CVaR(ES).

# 二.一致性风险度量
- ## Artzner等人在1999年提出了一个合理的风险度量应具备的条件————一致性风险度量公理化体系。
- ## 凡满足下列性质的风险度量$\rho$称为一致性风险度量：
- ## 1) 次可加性：$\forall R.V.(随机变量) X,Y,满足 \rho(X+Y) \leq \rho(X)+\rho(Y)$
- ## 2) 正齐次性：$\forall R.V.X,h>0,满足 \rho(hX) = h\rho(X)$
- ## 3) 平移不变性：$\forall R.V.X,a\in R,满足\rho(X+a) = \rho(X)+a$
- ## 4) 单调性：$\forall R.V. X,Y,X \leq Y,满足 \rho(X)\leq \rho(Y) $
- ## 从上述定义可以得知，我们常用的方差不满足平移不变性和单调性，而VaR不满足次可加性，所以它们都不是一致性风险度量
- ## 满足上述定义的风险度量方法也很多，比较经典的就是`CVaR和ES`

## 1.定义
- ## 条件风险价值CVaR(Conditional Value at Risk): 
![图片注释](http://storage-uqer.datayes.com/562f5464f9f06c4ca72fb74b/378a0654-bd25-11e6-b6a9-0242ac140002)
其中，$F_{\beta}=max([P(-X \leq x)-\beta]/(1-\beta),0),\beta=1-\alpha$
- ## 期望短缺ES(Expected Shortfall):
 
 $\qquad -(\int {X_{\alpha}du}) / {\alpha}$
 
 $其中，X_{\alpha}=inf \quad [ x|P(X\leq x) \geq\alpha ]$
 
- ## 由定义，可以得到，上述两种度量方法是等价的，并且满足一致性度量的定义
- ## 直观的讲，`ES度量的是损失在VaR水平之上的平均损失值`

## 2.计算
- ## 关于CVaR的计算方法也有很多种，这里介绍一种依据样本来计算的方法

```python
#加载需要使用的包
from __future__ import division
import numpy as np
import pandas as pd

from CAL.PyCAL import *    # CAL.PyCAL中包含font
font.set_size(22)
small_font = font.copy()
small_font.set_size(12)
figsize = (14,14*0.518)

import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('white')

from math import *
import scipy.optimize as opt
cal = Calendar('China.SSE')
```

- ## 举个例子，取沪深300成分股从20120101到20140101的日度收益率，计算每只股票的CVaR

```python
return_mat=DataAPI.MktEqudAdjGet(secID=set_universe('000300.ZICN'),beginDate='20120101',endDate='20140101', field=u"secID,tradeDate,closePrice")
X = return_mat.pivot(index='tradeDate',columns='secID', values='closePrice').pct_change().fillna(0.0)
```

- ## 计算方法为：对于每只股票，先得到n个收益样本的顺序统计量$X_{(1)},\cdots$,再计算下式
- $$-(\sum_{i=1}^{m} X_{(i)})/m,其中，m为小于等于n*\alpha的最大值$$
- ## 当n充分大时，该估计量就是几乎处处收敛到CVaR的，但我们在实际的过程中样本不可能太大，可以使用自助抽样即重复抽样的方法来估计

```python
def histCVaR(histSeries,n,alpha= 0.01):
    histSeries = pd.DataFrame(histSeries)
    histSeries.sort([histSeries.columns[0]],ascending=True,inplace=True)
    histSeries['count'] = np.arange(1,n+1)
    m = int(n*alpha)
    histSeries = histSeries[histSeries['count'] <= m]
    return -histSeries[histSeries.columns[0]].sum()/m
```

- ## 设置不同的alpha，观察一下频数分布

```python
fig = plt.figure(figsize = fsize)
pd.Series(map(lambda x: histCVaR(X[x],X.shape[0], 0.01), X.columns), index = X.columns).hist(label = 'alpha = 0.01', bins = 20)
pd.Series(map(lambda x: histCVaR(X[x],X.shape[0], 0.05), X.columns), index = X.columns).hist(label = 'alpha = 0.05', bins = 20)
pd.Series(map(lambda x: histCVaR(X[x],X.shape[0], 0.10), X.columns), index = X.columns).hist(label = 'alpha = 0.10', bins = 20)
plt.legend(loc = 'best')
plt.xlabel('CVaR')
plt.ylabel(u"频数分布", fontproperties = font)
plt.title(u"不同风险水平下各股票CVaR频数分布直方图", fontproperties = font)
```

- ## 大致可以看到ES是关于$\alpha$的单调非增连续函数，与定义是相符的
- ## 下面可以通过优化的方法来计算最小CVaR的投资组合

```python
def pfCVaR(weights):
    return histCVaR(pd.DataFrame(np.mat(X)*np.mat(weights).T, columns = ['portVaR']).portVaR,X.shape[0], 0.05)

cons = ({'type': 'eq', 'fun': lambda x: sum(x) - 1})
res = opt.minimize(pfCVaR, [1/X.shape[1]]*X.shape[1], constraints= cons,bounds= [(0,1)]*X.shape[1], method = 'SLSQP')
res['fun']
```

- ## 我们得到的最小CVaR的值为0.9%

```python
weight = pd.DataFrame()
weight['weight'] = res['x']
weight.index = X.columns
weight.T
```

- ## 可以看到，有些权重非常小，以后使用的时候，可以让约束的边界更小一些
- ## 下面定义一个类来方便研究和调整参数

```python
class CVaR(object):
    
    def __init__(self,alpha,begindate,middledate,enddate,train_frequency='daily',universe_name='000300.ZICN'):
        ###----传入参数----
        self.train_begin = begindate  #训练开始时间
        self.train_end = cal.adjustDate(middledate).strftime('%Y%m%d') #训练结束时间
        self.test_bengin = self.train_end      #结束时间
        self.test_end = enddate
        self.frequency = train_frequency  #数据频率，只要日度，周度，月度三个选择
        self.alpha = alpha
        self.universe = set_universe(universe_name,self.train_end)  #股票池
        
        ###----自定义参数----
        self.datelist = self.get_datelist()  #传入开始和结束日期，得到交易日列表list
        self.weight = self.get_weight()
        self.accumReturns = [0]
        self.indexs = [0]
        
    ###得到日期列表
    def get_datelist(self):   
        from CAL.PyCAL import *  #从dataapi调取交易日序列
        data=DataAPI.TradeCalGet(exchangeCD=u"XSHG",beginDate=self.train_begin,endDate=self.train_end,field=['calendarDate','isWeekEnd','isMonthEnd','isOpen'])
        if self.frequency == 'daily':   #按照规定的频率处理得到的交易日序列
            data = data[data['isOpen'] == 1]
        if self.frequency == 'weekly':
            data = data[data['isWeekEnd'] == 1]
        if self.frequency == 'monthly':
            data = data[data['isMonthEnd'] == 1]
        return data['calendarDate']
    
    
    #计算资产权重
    def get_weight(self):
        return_mat=DataAPI.MktEqudAdjGet(secID=self.universe,beginDate=self.train_begin,endDate=self.train_end,field=u"secID,tradeDate,closePrice")
        return_mat.set_index('tradeDate',inplace=True)
        return_mat = return_mat.ix[self.datelist,:]
        return_mat.reset_index(inplace=True)
        X = return_mat.pivot(index='tradeDate',columns='secID', values='closePrice').pct_change().fillna(0.0)
        alpha = self.alpha
        
        def histCVaR(histSeries,n,alpha):
            histSeries = pd.DataFrame(histSeries)
            histSeries.sort([histSeries.columns[0]],ascending=True,inplace=True)
            histSeries['count'] = np.arange(1,n+1)
            m = int(n*alpha)
            histSeries = histSeries[histSeries['count'] <= m]
            return -histSeries[histSeries.columns[0]].sum()/m
        
        def pfCVaR(weights):
            return histCVaR(pd.DataFrame(np.mat(X)*np.mat(weights).T, columns = ['portVaR']).portVaR,X.shape[0], alpha)

        cons = ({'type': 'eq', 'fun': lambda x: sum(x) - 1})
        res = opt.minimize(pfCVaR, [1/X.shape[1]]*X.shape[1], constraints= cons,bounds= [(0,1)]*X.shape[1], method = 'SLSQP')
        
        return res['x']
    
    
    
    #从日收益率序列得到累积收益率序列
    def get_accumReturns(self,daily):
        accumReturns=np.zeros(len(daily))
        for i in range(len(daily)):
            tmp = 1.0
            for j in range(i+1):
                tmp *= (1+daily[j])
            accumReturns[i] = tmp-1.0
        return accumReturns
    
    #得到年化收益率
    def get_annualReturns(self,accumReturns):
        annualReturns = (1+accumReturns[-1])**(250.0/len(accumReturns))-1
        return annualReturns
    
    #得到年波动率
    def get_volatility(self,daily):
        volatility = daily.std()*sqrt(250)
        return volatility
    
    #得到夏普比率
    def get_sharpe(self,annualReturns,volatility):
        rf=DataAPI.ChinaDataInterestRateInterbankRepoGet(indicID=u"M120000068",beginDate=self.test_bengin,\
                                                                         endDate=self.test_end,field=u"dataValue")
        rf = rf['dataValue'].values.tolist()
        rf = (rf[0]+rf[-1])/2/100/250         
        sharpe = (annualReturns-rf)/volatility
        return sharpe
    
    #得到最大回撤
    def get_maxDrawdown(self,accumReturns):
        drawdown = np.zeros(len(accumReturns))
        for i in range(1,len(drawdown)):
            drawdown[i] = 1.0-(1+accumReturns[i])/(1+max(accumReturns[:i]))
        maxDrawdown = max(drawdown)
        return maxDrawdown
    
    #得到Calmar比率
    def get_calmarRatio(self,annualReturns,maxDrawdown):
        calmarRatio = annualReturns/maxDrawdown
        return calmarRatio
    
    #定义一个计算投资组合风险指标的函数
    def get_result(self):
        return_mat=DataAPI.MktEqudAdjGet(secID=self.universe,beginDate=self.test_bengin, endDate=self.test_end, field=u"secID,tradeDate,closePrice")
        rtn_test = return_mat.pivot(index='tradeDate',columns='secID', values='closePrice').pct_change().fillna(0.0)
        
        Returns = np.mat(rtn_test.values)*np.mat(self.weight).T               #等权组合的收益率
        Returns = np.array(Returns).reshape(len(Returns),)       #把等权组合的收益率序列处理成数组

        #计算风险指标
        result2 = pd.DataFrame()
        accumReturns = self.get_accumReturns(Returns)    #调用函数计算各项指标
        annualReturns = self.get_annualReturns(accumReturns)
        volatility = self.get_volatility(Returns)
        sharpe = self.get_sharpe(annualReturns,volatility)
        maxDrawdown = self.get_maxDrawdown(accumReturns)
        calmarRatio = self.get_calmarRatio(annualReturns,maxDrawdown)
        result2['Markowitz组合'] = [accumReturns[-1],annualReturns,volatility,sharpe,maxDrawdown,calmarRatio]  #把结果存进dataframe里
        result2.index = ['累计收益','年化收益率','波动率','夏普比率','最大回撤','Calmar比率']
        self.indexs = rtn_test.index
        self.accumReturns = accumReturns
        return result2
    
    #画图
    def get_accumCurve(self):
        accumCurve = pd.DataFrame()
        accumCurve['Markowitz组合'] = self.accumReturns
        accumCurve.index = self.indexs
        ax = accumCurve.plot(figsize=figsize, title=u'累计收益曲线',linewidth=2.5)
        ax.legend([u'Markowitz组合'], prop=small_font)
        ax.title.set_font_properties(font)
        ax.set_axis_bgcolor('white')
        ax.grid(color='gray', alpha=0.2, axis='y')
        return ax
```

```python
test = CVaR(alpha=0.05,begindate='20120101',middledate='20140101',enddate='20150101',train_frequency='daily',universe_name='000300.ZICN')
```

- ## 可以看到，在我们的例子中，该方法能有效的控制风险和回撤，并保证一定程度的收益

```python
#计算一个序列的histCVaR
def histCVaR(histSeries,n,alpha):
    histSeries = pd.DataFrame(histSeries)
    histSeries.sort([histSeries.columns[0]],ascending=True,inplace=True)
    histSeries['count'] = np.arange(1,n+1)
    m = int(n*alpha)
    histSeries = histSeries[histSeries['count'] <= m]
    return -histSeries[histSeries.columns[0]].sum()/m
#计算一个投资组合的histCVaR
def pfCVaR(weights):
    return histCVaR(pd.DataFrame(np.mat(data)*np.mat(weights).T, columns = ['portVaR']).portVaR,data.shape[0], alpha)
#得到权重
def get_weight(data):
    cons = ({'type': 'eq', 'fun': lambda x: sum(x) - 1})
    res = opt.minimize(pfCVaR, [1/data.shape[1]]*data.shape[1], constraints= cons,bounds= [(0.001,_single_asset_capacity)]*data.shape[1], method = 'SLSQP')
    weight = pd.DataFrame()
    weight['weight'] = res['x']
    weight.index = data.columns
    return weight
```
