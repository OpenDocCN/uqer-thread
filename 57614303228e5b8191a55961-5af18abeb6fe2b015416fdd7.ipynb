第一个问题，~
下面试一试加波浪和不加波浪

```python
wts={'000001':0.1}
RSI = DataAPI.MktStockFactorsOneDayGet(tradeDate='20110104',secID='000001.XSHE',field=u"secID,RSI",pandas="1")
RSI['ticker'] = RSI['secID'].apply(lambda x: x[0:6])
RSI.set_index('ticker',inplace=True)
RSI['wts'] = np.nan
RSI['wts'][wts.keys()] = wts.values()
RSI = RSI[~np.isnan(RSI['wts'])]
print RSI
```

```python
wts={'000001':0.1}
RSI = DataAPI.MktStockFactorsOneDayGet(tradeDate='20110104',secID='000001.XSHE',field=u"secID,RSI",pandas="1")
RSI['ticker'] = RSI['secID'].apply(lambda x: x[0:6])
RSI.set_index('ticker',inplace=True)
RSI['wts'] = np.nan
RSI['wts'][wts.keys()] = wts.values()
RSI = RSI[np.isnan(RSI['wts'])]
print RSI
```

百度了一下，~是取反的意思，也就是说，如果不加~，取的就是表格里是NAN的值，加了~就是取不是nan的值。
写法好高级，果然是大神的代码，又膜拜一次。
