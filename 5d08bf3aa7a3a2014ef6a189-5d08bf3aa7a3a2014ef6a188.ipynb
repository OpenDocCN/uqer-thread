# 行业贝塔计算：加权与不加权

```python
###股票的行业贝塔（未加权）
sec=set_universe('A')
import pandas as pd
df=DataAPI.MktStockFactorsOneDayGet(tradeDate=u"20190416",secID=sec,field=u"ticker,HBETA",pandas="1")
df1=DataAPI.EquIndustryGet(secID=sec,industryVersionCD=u"010303",industry=u"",industryID=u"",industryID1=u"",
                       industryID2=u"",industryID3=u"",intoDate=u"20190416",field=u"ticker,industryName1",pandas="1")
pd.concat([df,df1],axis=1).dropna().groupby('industryName1').mean().sort_values('HBETA')
```

```python
##行业贝塔估值（加权平均）
import numpy as np
import pandas as pd
sec=set_universe('A')
df=DataAPI.MktStockFactorsOneDayGet(tradeDate=u"20190531",secID=sec,field=u"ticker,HBETA",pandas="1").dropna()
df1=DataAPI.MktEqudGet(secID=u"",ticker=u"",tradeDate=u"20190531",beginDate=u"",endDate=u"",isOpen="",field=u"ticker,marketValue",pandas="1")
df=df.merge(df1,left_on=['ticker'], right_on=['ticker'])
df2=DataAPI.EquIndustryGet(secID=sec,industryVersionCD=u"010303",industry=u"",industryID=u"",industryID1=u"",industryID2=u"",industryID3=u"",intoDate=u"",field="ticker,industryName1",pandas="1")
df=df.merge(df2,left_on=['ticker'], right_on=['ticker'], how='inner')
hybeta=pd.DataFrame(index=df['industryName1'].drop_duplicates(),columns=['beta'])
for a in df['industryName1'].drop_duplicates():
    df1=df[df['industryName1']==a]
    df1['weight']=df1['marketValue']/(df1['marketValue'].sum())
    df1['weightBETA']=df1['HBETA']*df1['weight']
    hybeta.loc[a]=df1['weightBETA'].sum().round(2)
print hybeta['beta'].sort_values()
```
