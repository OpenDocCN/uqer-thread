```python
#3.6pandas合并concat
#concatenating
import pandas as pd
import numpy as np

df1=pd.DataFrame(np.ones((3,4))*0,columns=['a','b','c','d'])
df2=pd.DataFrame(np.ones((3,4))*1,columns=['a','b','c','d'])
df3=pd.DataFrame(np.ones((3,4))*2,columns=['a','b','c','d'])
print(df1)
print(df2)
print(df3)

res=pd.concat([df1,df2,df3],axis=0)#竖向合并
print(res)
res1=pd.concat([df1,df2,df3],axis=0,ignore_index=True)#忽略掉原先的索引，得到的所有就是默认从0-8
print(res1)

#join,['inner','outer']
import pandas as pd
import numpy as np

df1=pd.DataFrame(np.ones((3,4))*0,columns=['a','b','c','d'],index=[1,2,3])
df2=pd.DataFrame(np.ones((3,4))*1,columns=['b','c','d','e'],index=[2,3,4])
print(df1)
print(df2)
res=pd.concat([df1,df2])#两者都有的部分全部呈列，只存在其中一个的用nan表示，即join=‘outer’
print(res)
res1=pd.concat([df1,df2],join='outer')
print(res1)
res2=pd.concat([df1,df2],join='inner',ignore_index=True)#只输出两者都有的部分
print(res2)

#join_axes
import pandas as pd
import numpy as np

df1=pd.DataFrame(np.ones((3,4))*0,columns=['a','b','c','d'],index=[1,2,3])
df2=pd.DataFrame(np.ones((3,4))*1,columns=['b','c','d','e'],index=[2,3,4])
res=pd.concat([df1,df2],axis=1)#简单左右合并
print(res)
res1=pd.concat([df1,df2],axis=1,join_axes=[df1.index])#根据df1的索引左右合并
print(res1)

#append
df1=pd.DataFrame(np.ones((3,4))*0,columns=['a','b','c','d'])
df2=pd.DataFrame(np.ones((3,4))*1,columns=['a','b','c','d'])
df3=pd.DataFrame(np.ones((3,4))*1,columns=['a','b','c','d'])
res=df1.append(df2,ignore_index=True)#竖向合并2个列表，忽略原先的索引
print(res)
res1=df1.append([df2,df3],ignore_index=True)
print(res1)

s1=pd.Series([1,2,3,4],index=['a','b','c','d'])#列表添加一个序列，相当于vstack功能
res2=df1.append(s1,ignore_index=True)
print(res2)

#3.7pandas合并merge
#merge two df by key/keys.(may be used in database)
import pandas as pd
import numpy as np
left=pd.DataFrame({'key':['K0','K1','K2','K3'],'A':['A0','A1','A2','A3'],'B':['B0','B1','B2','B3']})
right=pd.DataFrame({'key':['K0','K1','K2','K3'],'C':['C0','C1','C2','C3'],'D':['D0','D1','D2','D3']})
print(left)
print(right)
res=pd.merge(left,right,on='key')
print(res)

#consider two keys
import pandas as pd
import numpy as np
left=pd.DataFrame({'key1':['K0','K0','K1','K2'],'key2':['K0','K1','K0','K1'],'A':['A0','A1','A2','A3'],'B':['B0','B1','B2','B3']})
right=pd.DataFrame({'key1':['K0','K1','K1','K0'],'key2':['K0','K0','K0','K0'],'C':['C0','C1','C2','C3'],'D':['D0','D1','D2','D3']})
print(left)
print(right)
print'默认合并:'
res=pd.merge(left,right,on=['key1','key2'])
print(res)
#how=['left','right','inner','outer']默认的是inner
print'inner合并:'
res1=pd.merge(left,right,on=['key1','key2'],how='inner')
print(res1)
print'outer合并:'
res2=pd.merge(left,right,on=['key1','key2'],how='outer')
print(res2)
print'left合并:'
res3=pd.merge(left,right,on=['key1','key2'],how='left')
print(res3)
print'right合并:'
res4=pd.merge(left,right,on=['key1','key2'],how='right')
print(res4)

#indicator显示怎样merge
df1 = pd.DataFrame({'col1':[0,1], 'col_left':['a','b']})
df2 = pd.DataFrame({'col1':[1,2,2],'col_right':[2,2,2]})
print(df1)
print(df2)
res1=pd.merge(df1,df2,on='col1',how='outer',indicator=True)
print(res1)
res2=pd.merge(df1,df2,on='col1',how='outer',indicator='indicator_column')
#给一个自定义的名字give the indicator a custom name
print(res2)

#merge by index
left=pd.DataFrame({'A':['A0','A1','A2'],'B':['B0','B1','B2']},index=['K0','K1','K2'])
right=pd.DataFrame({'C':['C0','C1','C2'],'D':['D0','D1','D2']},index=['K0','K2','K3'])
print(left)
print(right)
res_inner=pd.merge(left,right,left_index=True,right_index=True,how='inner')
res_outer=pd.merge(left,right,left_index=True,right_index=True,how='outer')
res_left=pd.merge(left,right,left_index=True,right_index=True,how='left')
res_right=pd.merge(left,right,left_index=True,right_index=True,how='right')
print(res_inner)
print(res_outer)
print(res_left)
print(res_right)

#handle overlapping
boys=pd.DataFrame({'K':['k0','k1','k2'],'age':[1,2,3]})
girls=pd.DataFrame({'K':['k0','k0','k3'],'age':[4,5,6]})
print(boys)
print(girls)
res=pd.merge(boys,girls,on='K',suffixes=['_boys','_girls'],how='inner')
#overlap的时候为了区分两个表格中的age，加了前缀以作区分
print(res)

#3.8pandas plot数据可视化
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#plot methods:'bar','hist','box','kde','area','scatter','hexbin','pie'

#Series
data=pd.Series(np.random.randn(1000),index=np.arange(1000))
data=data.cumsum()#累加
data.plot()#显示图标
plt.show()#显示折线图

#DataFrame
data=pd.DataFrame(np.random.randn(1000,4),index=np.arange(1000),columns=list("ABCD"))
data=data.cumsum()
print(data.head())
data.plot()
plt.show()

#散点图scatter
ax=data.plot.scatter(x='A',y='B',color='DarkBlue',label='Class 1')#ax坐标系
data.plot.scatter(x='A',y='C',color='DarkGreen',label='Class 2',ax=ax)#ax指的是与上一张散点图在同一坐标系
plt.show()
```
