#### 首先是200个小市值不经过五日十日线筛选的策略（对照组）：

#### 然后是加了五日十日线的筛选后的上面哪个策略（实验组）：

```python
def buy_sale(account):
    hist = account.get_attribute_history('preClosePrice',21)
    close = account.get_attribute_history('closePrice',1)
    hist = pd.DataFrame(hist)
    close = pd.DataFrame(close) # dataframe, 按行排列
    
    m10_1 = hist.tail(10).mean()
    m10_2 = hist[-11:-1].mean()
    m10_3 = hist.head(10).mean()
    m5_1 = hist.tail(5).mean()
    m5_2 = hist[-6:-1].mean()
    
    close.ix['m10_1'] = pd.Series(m10_1,index=close.columns)
    close.ix['m10_2'] = pd.Series(m10_2,index=close.columns)
    close.ix['m10_3'] = pd.Series(m10_3,index=close.columns)
    close.ix['m5_1'] = pd.Series(m5_1,index=close.columns)
    close.ix['m5_2'] = pd.Series(m5_2,index=close.columns)
    
    data = close.T
    
    call = data[(data[0]>data.m5_1)&(data.m10_1>data.m10_2)&(data.m5_1>data.m5_2)&(data.m5_1>data.m10_1)]
    put = data[(data[0]<data.m5_1)&(data.m10_1<data.m10_2)&(data.m5_1<data.m5_2)&(data.m5_1<data.m10_1)&(data.m10_1<data.m10_3)]
    
    # put = data[(data[0]<data.m5_1*0.95) | (data[0]<data.m10_3*0.9) ]
    # put = data[ data[0]<data.m5_1*0.95 ]
               
    buy_list = list(call.index)
    put_list = list(put.index)
    
    return buy_list,put_list
```
