# 作业一：对比一下不同winsorize方法（正态分布去极值、分位数去极值）后的因子分布差异

```python
import numpy as np
from numpy.linalg import cholesky
import matplotlib.pyplot as plt

def winsorize_plot(data, vertical_lines=[]):
    '''
    data: Series形式，值为因子值
    '''
    fig = plt.figure()
    ax = fig.add_subplot(111)
    _df = ax.hist(data, 100, normed=True)
    color_list = ['#FF0000','#FF00FF','#EE00EE','#D02090','#ADFF2F','#7CFC00']
    count = 0
    for value in vertical_lines:
        ax.bar(value, 0.5, width=0.001, color=color_list[count], alpha=1)
        count += 1
    plt.show()
    
    

# 构造一批正态分布的数据
num = 3500
mean = 0.6
sigma = 1
np.random.seed(0)
factor_data = np.random.normal(mean, sigma, num)
factor_data = pd.Series(data=factor_data)
factor_data.index = [str(x).zfill(6) for x in factor_data.index.values]
# factor_data.head()



data_mean = factor_data.mean()
# 3倍标准差
std = np.std(factor_data)
std_up = data_mean+3*std
std_low = data_mean-3*std
print u'3倍标准差上下界', std_up, std_low

# 2.5%和97.5%分位数
quantile_low=np.percentile(factor_data, 2.5)
quantile_high = np.percentile(factor_data, 97.5)
print u'5%分位数上下界', quantile_low, quantile_high
winsorize_plot(factor_data, [std_up, std_low, quantile_low, quantile_high])
```

正负1倍标准偏差的概率 =68.3%
正负2倍标准偏差的概率 =95.5%
正负3倍标准偏差的概率 =99.7%

# 作业二：用python/pandas的基础功能自己计算spearman IC值以进一步体会spearman IC的实际意义

```python
import scipy.stats as st

# 用scipy的包计算
def calc_rank_ic(df):
    tmp_df = df.copy()
    tmp_df.dropna(inplace=True)
    return st.spearmanr(tmp_df['ROE'], tmp_df['return'])[0]

# 构造ROE数据
roe_data = np.random.rand(3000)
roe_data = [x-0.5 for x in roe_data]

# 构造return数据
return_data = np.random.randint(-100, 100, 3000)
return_data = [1.0*x/10 for x in return_data]

factor_data = pd.DataFrame({"ROE":roe_data, "return":return_data})
factor_data.head()
```

```python
calc_rank_ic(factor_data)
```

![图片注释](http://storage-uqer.datayes.com/560a3e12f9f06c597165ef9c/706339a4-5f28-11e8-889a-0242ac140002)

```python
# 第一步，将数值转化为排序值
factor_data['ROE_rank'] = factor_data['ROE'].rank()
factor_data['return_rank'] = factor_data['return'].rank()
factor_data.head()
```

```python
# 第二步，根据公式计算相关系数
mean_v = factor_data.mean()
mean_v
```

```python
# 计算各个值和均值的差
factor_data['ROE_rank_delta'] = factor_data['ROE_rank']-mean_v['ROE_rank']
factor_data['return_rank_delta'] = factor_data['return_rank']-mean_v['return_rank']

# 将偏差相乘，并累计求和
factor_data['delta_multip'] = factor_data['ROE_rank_delta']*factor_data['return_rank_delta']
up_value = factor_data['delta_multip'].sum()

# 再除以总个数，得到分母
up_value = up_value/len(factor_data)
up_value
```

```python
# 分子是两个的标准差相乘
down_value = factor_data['ROE_rank'].std()*factor_data['return_rank'].std()

# 分母/分子
v = up_value/down_value
v
```

```python
# 先取交易日历
calendar_frame = DataAPI.TradeCalGet(exchangeCD=u"XSHG,XSHE",beginDate=u"20100101",endDate=u"20171231",field=u"",pandas="1")
# 只取月末日期
calendar_frame = calendar_frame[calendar_frame.isMonthEnd==1]
calendar_frame['calendarDate'] = calendar_frame['calendarDate'].apply(lambda x: x.replace("-",""))

# 取流通市值数据
tfactor_frame = DataAPI.MktEqudAdjGet(secID=set_universe("ZZ500"),beginDate=u"20100101",endDate=u"20171231",isOpen="",field=['secID','tradeDate','negMarketValue','closePrice'],pandas="1")

tfactor_frame['tradeDate'] = tfactor_frame['tradeDate'].apply(lambda x: x.replace("-",""))
tfactor_frame = tfactor_frame[tfactor_frame.tradeDate.isin(calendar_frame.calendarDate.values)]
tfactor_frame.head()
```

```python
ic_calc_frame = tfactor_frame.copy()
def shift_data(df, n):
    df_copy = df.copy()
    df_copy.sort_values(by=['tradeDate'], ascending=True)
    df_copy['next_close'] = df_copy['closePrice'].shift(n)
    df_copy['return'] = (df_copy['next_close'] - df_copy['closePrice']) / df_copy['closePrice']
    return df_copy

# 收益往回移动一期，便于IC的计算
ic_calc_frame= ic_calc_frame.groupby(['secID']).apply(shift_data, -1)
ic_calc_frame.index=range(len(ic_calc_frame))
ic_calc_frame.head()
```

```python
# 用scipy的包计算
def calc_rank_ic(df):
    tmp_df = df.copy()
    tmp_df.dropna(inplace=True)
    return st.spearmanr(tmp_df['negMarketValue'], tmp_df['return'])[0]

ic_result = ic_calc_frame.groupby(['tradeDate']).apply(calc_rank_ic)
ic_result.mean()
```

```python
import time
from CAL.PyCAL import * 

start_time = time.time()
# ------------------------------------------------- 设定回测参数 ------------------------------------------------------

start = '2010-01-05'                       # 回测起始时间
end = '2017-12-31'                         # 回测结束时间
benchmark = 'HS300'                        # 策略参考标准
universe = DynamicUniverse('HS300')           # 证券池，支持股票和基金
capital_base = 10000000                     # 起始资金
freq = 'd'                              
refresh_rate = Monthly(1)  
accounts = {
    'fantasy_account': AccountConfig(account_type='security', capital_base=10000000)
}

# ------------------------------------------------- 读取因子数据 -------------------------------------------------------
factor_data = tfactor_frame[['secID', 'tradeDate', 'negMarketValue']]
factor_data.rename(columns={"negMarketValue":'factor'}, inplace=True)
factor_data = factor_data.set_index('tradeDate', drop=True)
factor_data.dropna(inplace=True)
q_dates = factor_data.index.values



# ------------------------------------------------ 将因子分组并回测 ---------------------------------------------------------

# 调用quick_backtest方法
# https://uqer.datayes.com/help/api/#quick_backtest参数优化
sim_params = quartz.SimulationParameters(start, end, benchmark, universe, capital_base, refresh_rate=refresh_rate, accounts=accounts)
# 获取回测行情数据
data = quartz.get_backtest_data(sim_params)
# 运行结果
results = {}

# 遍历每一组，进行回测
for quantile_five in range(1, 6):  
    def initialize(context):                   # 初始化虚拟账户状态
        pass

    def handle_data(context): 
        account = context.get_account('fantasy_account')
        current_universe = context.get_universe('stock', exclude_halt=True)
        pre_date = context.previous_date.strftime("%Y%m%d")
        if pre_date not in q_dates:            
            return

        # 拿取调仓日前一个交易日的因子，并按照相应分位选择股票
        q = factor_data.ix[pre_date].dropna()
        q = q.set_index('secID', drop=True)
        q = q.ix[current_universe]
        q.dropna(inplace=True)
        q_min = q['factor'].quantile((quantile_five-1)*0.2)
        q_max = q['factor'].quantile(quantile_five*0.2)
        my_univ = q[(q['factor']>=q_min) & (q['factor']<q_max)].index.values
        
       # 交易部分
        positions = account.get_positions()
        sell_list = [stk for stk in positions if stk not in my_univ]
        for stk in sell_list:
            order_to(stk,0)
        
        # 在目标股票池中的，等权买入
        for stk in my_univ:
            order_pct_to(stk, 1.0/len(my_univ))


    # 生成策略对象
    strategy = quartz.TradingStrategy(initialize, handle_data)
    
    # 开始回测, 返回说明见：https://uqer.datayes.com/help/api/#策略结果展示
    bt, perf = quartz.quick_backtest(sim_params, strategy, data=data)

    # 保存运行结果，1为因子最强组，5为因子最弱组
    results[6 - quantile_five] = {'max_drawdown': perf['max_drawdown'], 'sharpe': perf['sharpe'], 'alpha': perf['alpha'], 'beta': perf['beta'], 'information_ratio': perf['information_ratio'], 'annualized_return': perf['annualized_return'], 'bt': bt}    

    print ('backtesting for group %s..................................' % str(quantile_five)),
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style('white')

fig = plt.figure(figsize=(10,8))
fig.set_tight_layout(True)
ax1 = fig.add_subplot(211)
ax2 = fig.add_subplot(212)
ax1.grid()
ax2.grid()

for qt in results:
    bt = results[qt]['bt']

    data = bt[[u'tradeDate',u'portfolio_value',u'benchmark_return']]
    data['portfolio_return'] = data.portfolio_value/data.portfolio_value.shift(1) - 1.0   # 总头寸每日回报率
    data['portfolio_return'].ix[0] = data['portfolio_value'].ix[0]/	10000000.0 - 1.0
    data['excess_return'] = data.portfolio_return - data.benchmark_return                 # 总头寸每日超额回报率
    data['excess'] = data.excess_return + 1.0
    data['excess'] = data.excess.cumprod()                # 总头寸对冲指数后的净值序列
    data['portfolio'] = data.portfolio_return + 1.0     
    data['portfolio'] = data.portfolio.cumprod()          # 总头寸不对冲时的净值序列
    data['benchmark'] = data.benchmark_return + 1.0
    data['benchmark'] = data.benchmark.cumprod()          # benchmark的净值序列
    results[qt]['hedged_max_drawdown'] = max([1 - v/max(1, max(data['excess'][:i+1])) for i,v in enumerate(data['excess'])])  # 对冲后净值最大回撤
    results[qt]['hedged_volatility'] = np.std(data['excess_return'])*np.sqrt(252)
    results[qt]['hedged_annualized_return'] = (data['excess'].values[-1])**(252.0/len(data['excess'])) - 1.0
    ax1.plot(data['tradeDate'], data[['portfolio']], label=str(qt))
    ax2.plot(data['tradeDate'], data[['excess']], label=str(qt))
    

ax1.legend(loc=0)
ax2.legend(loc=0)
ax1.set_ylabel(u"净值", fontproperties=font, fontsize=16)
ax2.set_ylabel(u"对冲净值", fontproperties=font, fontsize=16)
ax1.set_title(u"因子不同五分位数分组选股净值走势", fontproperties=font, fontsize=16)
ax2.set_title(u"因子不同五分位数分组选股对冲后净值走势", fontproperties=font, fontsize=16)

# results 转换为 DataFrame
results_pd = pd.DataFrame(results).T.sort_index()

results_pd = results_pd[[u'alpha', u'beta', u'information_ratio', u'sharpe', u'annualized_return', u'max_drawdown',  
                         u'hedged_annualized_return', u'hedged_max_drawdown', u'hedged_volatility']]

cols = [(u'风险指标', u'Alpha'), (u'风险指标', u'Beta'), (u'风险指标', u'信息比率'), (u'风险指标', u'夏普比率'), (u'纯股票多头时', u'年化收益'),
        (u'纯股票多头时', u'最大回撤'), (u'对冲后', u'年化收益'), (u'对冲后', u'最大回撤'), (u'对冲后', u'收益波动率')]
results_pd.columns = pd.MultiIndex.from_tuples(cols)
results_pd.index.name = u'五分位组别'
results_pd
```
