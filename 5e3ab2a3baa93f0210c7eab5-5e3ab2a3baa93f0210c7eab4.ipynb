```python
#股票池筛选工具，移除ST、停牌、新股、市值处于尾部30%、银行和非银金融行业的股票
def Monthly_const_trade_calendar(beginDate=u"20050101",endDate=u""):
    """交易日历，可以选择cache，只调用 api 一次
    """
    data = DataAPI.TradeCalGet(exchangeCD=u"XSHG,XSHE", beginDate=beginDate, endDate=endDate, field=['calendarDate', 'isMonthEnd'])
    data = data[data.isMonthEnd == 1].calendarDate.tolist()
    data = sorted(list(set(data))) 
    return data

def st_remove(source_universe, st_date=None):
    """
    给定股票列表,去除其中在某日被标为ST的股票
    Args:
        source_universe (list of str): 需要进行筛选的股票列表
        st_date (datetime): 进行筛选的日期,默认为调用当天
    Returns:
        list: 去掉ST股票之后的股票列表

    Examples:
        >> universe = set_universe('A')
        >> universe_without_st = st_remove(universe)
    """
    st_date = st_date if st_date is not None else datetime.datetime.now().strftime('%Y%m%d')
    df_ST = DataAPI.SecSTGet(secID=source_universe, beginDate=st_date, endDate=st_date, field=['secID'])
    return [s for s in source_universe if s not in list(df_ST['secID'])]

#移除上市时间3个月内的股票
def new_remove(ticker,tradeDate= None,day = 75):
    tradeDate = tradeDate if tradeDate is not None else datetime.datetime.now()
    period = '-' + str(day) + 'B'
    cal = Calendar('China.SSE')
    pastDate = cal.advanceDate(tradeDate,period)
    pastDate = pastDate.strftime("%Y-%m-%d")

    tickerDist={}
    tickerShort=[] 
    for index in range(len(ticker)):
        OneTickerShort=ticker[index][0:6]
        tickerShort.append(OneTickerShort)
        tickerDist[OneTickerShort]=ticker[index]

    ipo_date = DataAPI.SecIDGet(partyID=u"",assetClass=u"",ticker=tickerShort,cnSpell=u"",field=u"ticker,listDate",pandas="1")
    remove_list = ipo_date[ipo_date['listDate'] > pastDate]['ticker'].tolist()
    remove_list=[values for keys,values in tickerDist.items() if keys in remove_list ] 
    return [stk for stk in ticker if stk not in remove_list] 

# 去除市值位于尾部30%的股票
def value_remove(source_universe,tradeDate= None):
    tradeDate = tradeDate if tradeDate is not None else datetime.datetime.now().strftime('%Y%m%d')
    mkt_value=DataAPI.MktEqudGet(secID=source_universe,ticker=u"",tradeDate=tradeDate,beginDate=u"",endDate=u"",isOpen="",field=u"secID,negMarketValue",pandas="1")
    mkt_value=mkt_value.sort('negMarketValue',ascending=True) 
    hang=int(round(len(mkt_value)*0.3))
    newmkt_value=mkt_value.head(hang)
    return [stk for stk in source_universe if stk not in list(newmkt_value['secID'])]

#去除银行和非银金融
def jinrong_remove(source_universe,tradeDate= None):
    tradeDate = tradeDate if tradeDate is not None else datetime.datetime.now().strftime('%Y%m%d')
    columns=['secID','tradeDate']
    new_value=DataAPI.MktEqudGet(secID=source_universe,ticker=u"",tradeDate=tradeDate,beginDate=u"",endDate=u"",isOpen="",field=columns,pandas="1")
#new_value.head()
#取出当日申万分级
    columns2=['secID','industryName1','isNew']
    type_new=DataAPI.EquIndustryGet(industry=u"",secID=source_universe,industryVersionCD=u"010303",intoDate=tradeDate,field=columns2,pandas="1")
    type_new=type_new[type_new['isNew']==1]
#将两者进行合并
    value_merge=pd.merge(new_value,type_new,on='secID')
    return [stk for stk in source_universe if stk not in  list(value_merge[value_merge.industryName1.isin(['银行','非银金融'])]['secID'])] 

#不知道平安银行为什么移除不了，再定义和函数移除一下
def pingan_remove(source_universe, st_date=None):
    st_date = st_date if st_date is not None else datetime.datetime.now().strftime('%Y%m%d')
    df_ST = DataAPI.MktEqudGet(secID=u"000001.XSHE",ticker=u"",tradeDate=st_date,beginDate=u"",endDate=u"",isOpen="",field=u"",pandas="1")
    return [s for s in source_universe if s not in list(df_ST['secID'])]

 

#对信号进行预处理，主要是进行去极值，标准化、中性化
def signalPre(factor,date):
    return standardize(neutralize(winsorize(factor), date))

#获取指定日期的前一年日期
def previousYearDate(strDate):
    try:
        date = datetime.datetime.strptime(strDate,'%Y%m%d')
    except ValueError as e:
        date = datetime.datetime.strptime(strDate,'%Y-%m-%d')
    previousYear = date.year-1
    previousMonth = date.month
    previousDay = date.day
    try:
        previousDate = datetime.datetime(previousYear,previousMonth,previousDay)
    except ValueError as e:
        previousDate = datetime.datetime(previousYear,previousMonth,previousDay-1)
    previousDate = previousDate.strftime('%Y%m%d')
    return previousDate

#如果当前日期是交易日，返回当前日期，若不是，则返回上一交易日
def getNearestTradeDay(strDate):
    cal_dates = DataAPI.TradeCalGet(exchangeCD=u"XSHG",beginDate=strDate, endDate=strDate).sort('calendarDate')
    cal_date = cal_dates.loc[0]
    if cal_date['isOpen'] == 1:
        return strDate
    else:
        return timeFormate(cal_date['prevTradeDate'])

def timeFormate(strDate):
    return datetime.datetime.strptime(strDate,'%Y-%m-%d').strftime('%Y%m%d')
#施密特正交化函数，输入n个向量的dataframe，输出n个向量的dataframe
def Schmidt(data):
    output = pd.DataFrame()
    mat = np.mat(data)
    output[0] = np.array(mat[:,0].reshape(len(data),))[0]
    for i in range(1,data.shape[1]):
        tmp = np.zeros(len(data))
        for j in range(i):
            up = np.array((mat[:,i].reshape(1,len(data)))*(np.mat(output[j]).reshape(len(data),1)))[0][0]
            down = np.array((np.mat(output[j]).reshape(1,len(data)))*(np.mat(output[j]).reshape(len(data),1)))[0][0]
            tmp = tmp+up*1.0/down*(np.array(output[j]))
        output[i] = np.array(mat[:,i].reshape(len(data),))[0]-np.array(tmp)
    output.index = data.index
    output.columns = data.columns
    return output
```

```python
trade_calendar = Monthly_const_trade_calendar(beginDate='20100101',endDate='20191130')
allFactor = DataFrame()
for date in trade_calendar:
    # 获取当日股票池
    univ=set_universe('A',date)
    univ=value_remove(univ,date)
    univ=jinrong_remove(univ,date)
    univ=pingan_remove(univ,date)
    univ=st_remove(univ,date)
    univ=new_remove(univ,date)
    
    factor = DataAPI.MktStockFactorsOneDayGet(tradeDate=date,secID=univ,ticker=u"",field=u"secID,tradeDate,"+'KDJ_K',pandas="1")           #获取原始数据
    factor.columns = ['secID','tradeDate','value']
    factor.dropna(inplace=True)
    factor = factor.loc[(factor['value']!=np.inf) &(factor['value']!=-np.inf)] #去掉正负无穷值
    factor.set_index('secID',inplace=True)
    if len(factor) <= 1:
        continue
    factor['value'] = signalPre(factor['value'],date)          #对因子进行去极值，中性化，标准化
    factor = factor.reset_index()
    allFactor = allFactor.append(factor,ignore_index=True)
allFactor.to_csv('KDJ_K.csv')
```

```python
# 5.2 IC加权，取因子过去一段时间的IC的均值为权重
#思路：
# a、获取每一股票每一交易日的因子值，上面的第3步
# b、计算单个因子每一交易日的IC（截面上的IC：计算所有股票某一交易日因子与收益率的相关性），对取得的因子值向下偏移1期
# c、进行12期的窗口滚动求均值，偏移一期是为了用前面12期因子IC均值来预测当天IC，并用预测值来配因子的权重
# d、各因子的IC的dataframe进行横向合并，并在每一交易日上求得每一因子IC的权重（每一因子IC值除以所有因子IC绝对值之和）
# e、把各因子乘以权重进行求和得到复合因子

import pandas as pd
import numpy as np
import scipy.stats as st
def get_rank_ic(factor):
    """
    计算因子的信息系数
    输入：
        factor:DataFrame，index为日期，columns为股票代码，value为因子值
        forward_return:DataFrame，index为日期，columns为股票代码，value为下一期的股票收益率
    返回：
        DataFrame:index为日期，columns为IC，IC t检验的pvalue
    注意：factor与forward_return的index及columns应保持一致
    """
    forward_1m_return_data = pd.read_csv('e_Ret_1M(shizhi).csv').dropna()    # 未来1个月收益率 
    forward_1m_return_data=forward_1m_return_data.pivot(index='tradeDate',columns='secID',values='e_return')
    forward_1m_return_data.index = pd.to_datetime(forward_1m_return_data.index)
   
    nMV = pd.read_csv('negMarketValue.csv').dropna()
    nMV = nMV.pivot(index='tradeDate',columns='secID',values='negMarketValue')
    nMV.index = pd.to_datetime(nMV.index)
    
    common_index = factor.index.intersection(forward_1m_return_data.index)#intersection获得索引的交集，共同的交易日期
    ic_data = pd.DataFrame(index=common_index, columns=['IC','pValue'])#创建一个空的dataframe
    
    
    # 计算相关系数
    for dt in ic_data.index:
        tmp_factor = factor.ix[dt]
        tmp_ret = forward_1m_return_data.ix[dt]
        tmp_nwv=nMV.ix[dt]
        cor = pd.DataFrame(tmp_factor)
        ret = pd.DataFrame(tmp_ret)
        nwv=pd.DataFrame(tmp_nwv)
        cor.columns = ['value']
        ret.columns = ['e_return']
        nwv.columns = ['negMarketValue']
        cor['e_return'] = ret['e_return']
        cor['negMarketValue']=nwv['negMarketValue']
        cor = cor[~pd.isnull(cor['value'])][~pd.isnull(cor['e_return'])]
        if len(cor) < 5:
            continue
     # WLS回归
        X=cor['value']
        y=cor['e_return']
        wls = sm.WLS(y, X, weights=cor['negMarketValue'])
        result = wls.fit()
        cor['Residual'] = result.resid
        ic,pvalue = st.spearmanr(cor['value'],cor['Residual'])   # 计算秩相关系数 RankIC
        ic_data['IC'][dt] = ic
    return ic_data

# b、计算单个因子每一交易日的IC（截面上的IC：计算所有股票某一交易日因子与收益率的相关性），对取得的因子值向下偏移1期
#EP_ic = get_rank_ic(EP)
#print len(EP_ic)#总的月末交易日为98，pe_ic长度也为98，说明没有那一天数据量小于5
EP_ic = get_rank_ic(EP).shift(1)#向下移一行
BP_ic = get_rank_ic(BP).shift(1)
SP_ic = get_rank_ic(SP).shift(1)
CFP_ic = get_rank_ic(CFP).shift(1)
TA2EV_ic = get_rank_ic(TA2EV).shift(1)
CTOP_ic = get_rank_ic(CTOP).shift(1)
# c、进行12期的窗口滚动求均值，偏移一期是为了用前面12期因子IC均值来预测当天IC，并用预测值来配因子的权重
EP_rolling_ic = EP_ic.rolling(12).mean()
BP_rolling_ic = BP_ic.rolling(12).mean()
SP_rolling_ic = SP_ic.rolling(12).mean()
CFP_rolling_ic = CFP_ic.rolling(12).mean()
TA2EV_rolling_ic = TA2EV_ic.rolling(12).mean()
CTOP_rolling_ic = CTOP_ic.rolling(12).mean()
# d、各因子的IC的dataframe进行横向合并，并在每一交易日上求得每一因子IC的权重（每一因子IC值除以所有因子IC绝对值之和）
all_rolling_ic_df = pd.concat((EP_rolling_ic['IC'], BP_rolling_ic['IC'],SP_rolling_ic['IC'],CFP_rolling_ic['IC'],TA2EV_rolling_ic['IC'],CTOP_rolling_ic['IC']), axis=1)#对pe和pb进行横向拼接
all_rolling_ic_df.columns = ['EP', 'BP','SP','CFP','TA2EV','CTOP']#更改列名
factor_weight = all_rolling_ic_df.divide(all_rolling_ic_df.abs().sum(axis=1), axis=0)#dataframe在列向的除法，得到每一个因子的权重

# e、把各因子乘以权重进行求和得到复合因子
factor = EP.multiply(factor_weight['EP'], axis=0) + BP.multiply(factor_weight['BP'], axis=0)+SP.multiply(factor_weight['SP'], axis=0)+CFP.multiply(factor_weight['CFP'], axis=0)+TA2EV.multiply(factor_weight['TA2EV'], axis=0)+CTOP.multiply(factor_weight['CTOP'], axis=0)
#factor=factor.unpivot(index='tradeDate', columns='secID', values='value')
factor.to_csv('ic_guzhi.csv')
#factor.head(30)
```

```python
# 可编辑部分与 strategy 模式一样，其余部分按本例代码编写即可
import pandas as pd
# -----------回测参数部分开始，可编辑------------
start = '2018-01-01'                       # 回测起始时间
end = '2019-12-31'                         # 回测结束时间
benchmark = 'ZZ500'                        # 策略参考标准
universe = set_universe('A')               # 证券池，支持股票和基金
capital_base = 10000000                    # 起始资金
freq = 'd'                                 # 策略类型，'d'表示日间策略使用日线回测
refresh_rate = 1                           # 调仓频率，表示执行handle_data的时间间隔

factor_data = pd.read_csv('dengquan_guzhidongliang.csv')     # 读取量价因子数据
factor_data = factor_data[['tradeDate','secID','value']]
factor_data = factor_data.pivot(index='tradeDate',columns='secID',values='value')
factor_dates = factor_data.index.values
# ---------------回测参数部分结束----------------


# 把回测参数封装到 SimulationParameters 中，供 quick_backtest 使用
sim_params = quartz.SimulationParameters(start, end, benchmark, universe, capital_base)
# 获取回测行情数据
idxmap, data = quartz.get_daily_data(sim_params)
# 运行结果
results_corr = {}

# 调整参数(选取股票的量价因子五分位数)，进行快速回测
for quantile_five in range(1, 6):
    
    # ---------------策略逻辑部分----------------
    commission = Commission(0.0002,0.0002)      # 交易费率设为双边万分之二

    def initialize(account):                   # 初始化虚拟账户状态
        pass

    def handle_data(account):                  # 每个交易日的买入卖出指令
        pre_date = account.previous_date.strftime("%Y-%m-%d")
        if pre_date not in factor_dates:            # 只在计算过量价因子的交易日调仓
            return

        # 拿取调仓日前一个交易日的量价因子，并按照相应五分位选择股票
        pre_factor = factor_data.ix[pre_date]
        pre_factor = pre_factor.dropna()

        pre_factor_min = pre_factor.quantile((quantile_five-1)*0.2)
        pre_factor_max = pre_factor.quantile(quantile_five*0.2)
        my_univ = pre_factor[pre_factor>=pre_factor_min][pre_factor<pre_factor_max].index.values

        # 调仓逻辑
        univ = [x for x in my_univ if x in account.universe]

        # 不在股票池中的，清仓
        for stk in account.security_position:
            if stk not in univ:
                account.order_to(stk, 0)
        # 在目标股票池中的，等权买入
        for stk in univ:
            account.order_pct_to(stk, 1.1/len(univ))
    # ---------------策略逻辑部分结束----------------

    # 把回测逻辑封装到 TradingStrategy 中，供 quick_backtest 使用
    strategy = quartz.TradingStrategy(initialize, handle_data)
    # 回测部分
    bt, acct = quartz.quick_backtest(sim_params, strategy, idxmap, data, refresh_rate=refresh_rate, commission=commission)

    # 对于回测的结果，可以通过 perf_parse 函数计算风险指标
    perf = quartz.perf_parse(bt, acct)

    # 保存运行结果
    tmp = {}
    tmp['bt'] = bt
    tmp['annualized_return'] = perf['annualized_return']
    tmp['volatility'] = perf['volatility']
    tmp['max_drawdown'] = perf['max_drawdown']
    tmp['alpha'] = perf['alpha']
    tmp['beta'] = perf['beta']
    tmp['sharpe'] = perf['sharpe']
    tmp['information_ratio'] = perf['information_ratio']
    
    results_corr[quantile_five] = tmp
    print str(quantile_five),
print 'done'
```
