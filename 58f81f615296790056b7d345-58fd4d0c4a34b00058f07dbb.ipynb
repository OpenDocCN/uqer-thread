# AdaBoost选股策略

引言
====
本策略的主要思路是根据股票的下期收益将其做一个二元分类，将股票分为预期表现好的和预期表现差的，由于优矿平台不允许做空，我们就只买入模型预测表现好的那些股票，在下次预测开始前卖出。

AdaBoost算法简介
===============
处理分类问题，很容易想到求助于机器学习算法，诸如支持向量机,神经网络等。而这些算法都有一个明显的缺点，就是容易过拟合，且对输入数据敏感，尽管人们在算法本身的框架内想出了各种办法来消除这些问题，结果却仍然难以让人满意。而在1996年由R.Scharpire和Y.Freund提出的AdaBoost算法（Adaptive Boosting Algorithm），带给我们一种全新的思路来重新看待这些问题，AdaBoost算法属于更广泛的一类叫做系综方法的一种，此后，机器学习中系综方法的研究如雨后春笋般展开。
    AdaBoost算法严格说来并不算是一种分类算法，它只是把一系列的弱分类算法整合起来，最终得到一个强的分类算法。这里的强和弱是通过算法错误分类训练数据的数量来衡量。通过不断调整数据的权重，不断改进我们的算法，最终实现强的分类算法。具体说来就是对每一个弱分类器，分类完成后，我们找出哪些是正确分类以及哪些被错误分类。我们降低那些正确分类的数据的权重，对被错误分类的那些数据做相反的处理，这样我们就能够把精力更多地放在那些分类错误的数据上，从而得到更精细的分类。
    由于AdaBoost算法通常选择一些非常简单的分类方法做为弱分类器，例如经典的决策树桩（decision stumps），使得整个算法的容错性更好，从而过拟合问题不再突出。而且得益学者们广泛而又深入地理论研究，人们对AdaBoost算法有更好的理解和掌控。当然AdaBoost算法也有自己的缺陷，比如难以确定弱分类器的最优数目以及同样地对输入数据的敏感。

策略简介
=======
在我们的模型中,我们用每一个因子生成一个弱分类器，然后利用AdaBoost算法将它们有机地整合。具体过程如下：
1.对一股票数据的集合$S=\\{(x^{(1)},y^{(1)}),...,(x^{(m)},y^{(m)})\\}$，$m$是股票的数量，$x^{(i)}$是第$i$支股票的因子值。这里注意，在实际操作过程中，我们并没有直接采用股票因子值的原始数据，而是将其转换为根据排名得到的相对值。这是因为机器学习算法对训练数据相当敏感，股票因子值在整体中的排名更具有稳健性。$y^{(i)}$标记该股票在未来一个月的收益，对于集合$S$中的股票，收益前30%的股票，$y^{(i)}=1$，收益后30%的股票，$y^{(i)}=-1$。中间的数据我们视为噪音，弃之不用。
2.在我们的模型中，一个弱分类器$h$由因子库$F$中的一个因子$k$确立，是一个从因子空间到分类信心空间的一个函数。对因子$k$和股票$i$，其因子值记为$x^{(i)}_k$，我们赋予权重$w^{(i)}_{k}$，我们设置$w^{(i)}_0=1/m$。如果第$k$个因子值落在第$j$分段，我们定义弱分类器$h_k$为：
\begin{equation}
h^{(i)}_k=\frac12\log(\frac{W^j_{k+}+\epsilon}{W^j_{k-}+\epsilon}),
\end{equation}
其中$\epsilon=1/m$，$j=1,2,...,Q$，$Q$是分的类数（这里我们取$Q=2$)，并且$W^{j}_{k\pm}$的定义为：
\begin{equation}
W^{j}_{k\pm}=\sum\limits_{y^{(i)}=\pm 1\atop x^{(i)}_k\in\text{quantile}~j}w^{(i)}_k
\end{equation}
直觉上来讲，如果一个分类中表现好的权重越多，那么 $h_k$的值将更大。来如果某个股票的因子落在这个分类中，那么该股票将更有可能表现好，反之，如果 $h_k$的值越小，落在该分类中的股票的未来表现有可能为差。
3.我们定义一个函数$Z_k$，用来衡量由因子$k$所确立的弱分类器的好坏：
\begin{equation}
Z_k=\sum_{j=1}^{Q}\sqrt{W^{j}_{k+}W^{j}_{k-}}
\end{equation}
在每一个分类中，因为所有的权重和为1。因此若$W^{j}_{k+}$和$W^{j}_{k-}$的区别较大时，函数$Z_k$的数值则小。
在每一轮过后，我们都要更新每支股票的权重：
\begin{equation}
w^{(i)}_{t+1}=w^{(i)}_{t}\exp{(-y^{(i)}h^{(i)}_{t})},
\end{equation}
$t$是我们进行boosting操作的次数。
4.当$t$跑遍我们选取的因子库时，我们得到最终强的分类器$h$：
\begin{equation}
h^{(i)}=\sum_{k=1}^{|F|}h^{(i)}_{k}
\end{equation}
5.对于我们这里的情形，我们对股票在未来一个月表现的预测为：
\begin{equation}
\hat{y}^{(i)}=\text{sign}(h^{(i)})
\end{equation}

代码实现
=======

```python
import numpy as np
import pandas as pd
from CAL.PyCAL import *
#因子库列表
factors=['BIAS20','RSI','ARTDays','InventoryTDays','DebtsAssetRatio','NetProfitRatio','CurrentRatio','QuickRatio','ROE','ROA','PS','KlingerOscillator','PE', 'NetProfitGrowRate', 'MA10', 'MA60', 'LCAP', 'LFLO', 'NPToTOR', 'OperatingProfitGrowRate', 'TotalAssetGrowRate','DAREC','GREC','DEGM'] 
#获取某交易日的股票列表
def get_stock(day):       
    universe = DataAPI.IdxConsGet(ticker="000300",intoDate=day,field=['consID'])
    s=universe['consID'].values.tolist()
    stocks = [str(x) for x in s]
    return stocks
#获取某交易日股票的因子数据
def get_factor(day):
    cal = Calendar('China.SSE')
    period = Period('1M')
    day_after=cal.advanceDate(day,period,endOfMonth=True).toDateTime().strftime('%Y%m%d')
    stocks=get_stock(day)
    factor_data=DataAPI.MktStockFactorsOneDayGet(tradeDate=day,secID=stocks,field=['secID','tradeDate']+factors)
    factor_data.iloc[:,2:]=factor_data.iloc[:,2:].rank(method='first',ascending=False).apply(lambda x:x/len(factor_data))
    close_price=DataAPI.MktEqudGet(tradeDate=day,secID=stocks,field=['ticker','closePrice'])
    close_price_after=DataAPI.MktEqudGet(tradeDate=day_after,secID=stocks,field=['ticker','closePrice'])
    ret=(close_price_after['closePrice']-close_price['closePrice'])/close_price['closePrice']
    factor_data['return']=ret
    factor_data=factor_data.dropna()
    factor_data=factor_data.sort(columns='return',ascending=False)
    factor_data=factor_data.set_index(np.arange(len(factor_data)))
    return factor_data
#准备训练数据，按照股票在未来一个月的收益表现标记每一支股票
def get_raw_data(day):
    factor_data=get_factor(day)
    label_1=factor_data.ix[:int(0.3*len(factor_data))-1,:]
    label_0=factor_data.ix[len(factor_data)-int(0.3*len(factor_data)):,:]
    label_1['label']=1
    label_0['label']=-1
    raw_data=pd.concat([label_1,label_0],axis=0).drop('return',axis=1)
    raw_data['weight']=1.0/(len(raw_data))*np.ones(int(len(raw_data)))
    raw_data=raw_data.set_index(np.arange(len(raw_data)))
    return raw_data
```

```python
#定义AdaBoost函数,功能是完成上文提到的第2到第4步
def AdaBoost(raw_data):
    global factors
    while factors!=list():
        Z=list()
        W=list()
        #对于因子库中的每个因子,计算Z函数
        for f in factors:
            w={'W1p':0,'W1m':0,'W2p':0,'W2m':0}
            data=raw_data.sort(columns=f,ascending=False).set_index(np.arange(len(raw_data)))
            for i in range(int(0.5*len(data))):
                if data.ix[i,:]['label']==1:
                    w['W1p']+=data.ix[i,:]['weight']
                else:
                    w['W1m']+=data.ix[i,:]['weight']
            for i in range(int(0.5*len(data)),len(data)):
                if data.ix[i,:]['label']==1:
                    w['W2p']+=data.ix[i,:]['weight']
                else:
                    w['W2m']+=data.ix[i,:]['weight']
            z=np.sqrt(w['W1p']*w['W1m'])+np.sqrt(w['W2p']*w['W2m'])
            Z.append(z)
            W.append(w)
        #找到最小的Z值所对应的那个因子
        loc=Z.index(min(Z))
        raw_data=raw_data.sort(columns=factors[loc],ascending=False)
        #建立弱分类器
        h1=0.5*np.log((W[loc]['W1p']+1.0/len(raw_data))/(W[loc]['W1m']+1.0/len(raw_data)))
        h2=0.5*np.log((W[loc]['W2p']+1.0/len(raw_data))/(W[loc]['W2m']+1.0/len(raw_data)))
        h1=h1*np.ones(int(0.5*len(raw_data)))
        h2=h2*np.ones(len(raw_data)-int(0.5*len(raw_data)))
        weak_classfier=np.concatenate((h1,h2))
        raw_data['weakClassifier_'+factors[loc]]=weak_classfier
        for i in range(len(raw_data)):
            raw_data.ix[i,'weight']*=np.exp(-raw_data.ix[i,'label']*raw_data.ix[i,'weakClassifier_'+factors[loc]])
        raw_data['weight']=raw_data['weight']/raw_data['weight'].sum()
        factors.pop(loc)
    factors=['BIAS20','RSI','ARTDays','InventoryTDays','DebtsAssetRatio','NetProfitRatio','CurrentRatio','QuickRatio','ROE','ROA','PS','KlingerOscillator','PE', 'NetProfitGrowRate', 'MA10', 'MA60', 'LCAP', 'LFLO', 'NPToTOR', 'OperatingProfitGrowRate', 'TotalAssetGrowRate','DAREC','GREC','DEGM']
    #Boosting成强分类器
    raw_data=raw_data.set_index(np.arange(len(raw_data)))
    strong_classifier=list()
    for i in range(len(raw_data)):
        s=raw_data.ix[:,len(factors)+4:].loc[i].sum()
        strong_classifier.append(s)
    raw_data['strongClassifier']=np.array(strong_classifier)
    raw_data['train_label']=raw_data['strongClassifier'].apply(lambda x: 1 if x>0 else -1)
    trained_data=raw_data
    loss=0.5*np.abs(trained_data['label']-trained_data['train_label']).sum()
    return trained_data
```

```python
#预测
def predict(day):
    global factors
    cal = Calendar('China.SSE')
    period = Period('-1M')
    day_before=cal.advanceDate(day,period,endOfMonth=True).toDateTime().strftime('%Y%m%d')
    train_data=get_raw_data(day_before)
    trained_data=AdaBoost(train_data)
    stocks=get_stock(day)
    factor_data=DataAPI.MktStockFactorsOneDayGet(tradeDate=day,secID=stocks,field=['secID','tradeDate']+factors)
    factor_data=factor_data.dropna()
    factor_data=factor_data.set_index(np.arange(len(factor_data)))
    for f in factors:
        factor_data=factor_data.sort(columns=f)
        h1=trained_data.sort(columns=f,ascending=False)['weakClassifier_'+f].tolist()[0]
        h2=trained_data.sort(columns=f,ascending=False)['weakClassifier_'+f].tolist()[-1]
        h1=h1*np.ones(int(0.5*len(factor_data)))
        h2=h2*np.ones(len(factor_data)-int(0.5*len(factor_data)))
        weak_classfier=np.concatenate((h1,h2))
        factor_data['weakClassifier_'+f]=weak_classfier
        factor_data=factor_data.set_index(np.arange(len(factor_data)))
    strong_classifier=list()
    for i in range(len(factor_data)):
        s=factor_data.ix[:,len(factors)+2:].loc[i].sum()
        strong_classifier.append(s)
    factor_data['strongClassifier']=np.array(strong_classifier)
    factor_data['predict']=factor_data['strongClassifier'].apply(lambda x: 1 if x>0 else -1)
    return factor_data[['secID','predict']]
```
