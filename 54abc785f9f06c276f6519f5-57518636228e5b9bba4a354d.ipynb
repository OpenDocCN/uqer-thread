# 调用API的因子数据，生成适用于RDP的信号文件

此代码完成的功能是：

>* 获取沪深300成分股在一段时间内的PE值

>* 把这些PE数据按照天存储为csv文件

可以将这些文件下载到本地，打包成一个文件夹，上传到RDP（[通联策略研究](https://gw.wmcloud.com/rdp//#/signalMgr)）中当做信号使用。

```python
from datetime import timedelta
start = datetime(2014, 5, 1)				# 回测起始时间
end   = datetime(2014, 7, 1)				# 回测结束时间
benchmark = 'HS300'							# 策略参考标准
universe = set_universe('HS300')	        # 股票池
capital_base = 100000                       # 起始资金
num = len(universe)/50
tickers=[]    #ticker储存的股票代码没有XSHG,XSHE的后缀，符合DataAPI的格式要求
for stk in universe:
    tickers.append(stk[0:6])
cal = Calendar('China.SSE')    #调用CAL下的‘上海证券交易所交易日历’
today = start
while((today-end).days<0):    #如果不是最后一天
    today_CAL = Date.fromDateTime(today)    #转换为CAL下的日期格式
    if(cal.isBizDay(today_CAL)):    #判断是否是交易日，如果今天是交易日，则调取当日各股票的PE值
        today_str=today.strftime("%Y%m%d")    #today的格式本为datetime，这里换成字符串格式，符合DataAPI中的时间格式要求
        info = DataAPI.MktStockFactorsOneDayGet (stockID=tickers,date=today_str,field='PE',pandas='1')    #调用DataAPI中有关PE的函数
        info = info[['ticker','PE']]    #只保留ticker和PE两项
        file_name = today_str + '.csv'
        info.to_csv(file_name,index=False,header=False)
    today = today+timedelta(days=1)    #下一天
```
