```python
import pandas as pd
```

```python
# raw data from report 《九泰锐智定增灵活配置混合型证券投资基金2016年半年度报告》
df = pd.DataFrame({
    'secID':['002709','002460','603601','603368','300017','300370','002706','002425'],
    'date':['2015-11-11','2015-09-29','2016-05-16','2016-02-19','2016-03-11','2015-11-05','2016-04-06','2016-05-06'],
    'subscription':[14.00,12.08,30.00,55.28,43.95,6.15,17.99,21.57],
    'volume':[2000000,2483442,1000000,542692,682594,3252032,1417848,1010663]
    })
df['market_price'] = 0
df['cost'] = df['subscription'] * df['volume']
df['secShortName'] = df['secID'].apply(lambda x: DataAPI.SecIDGet(partyID=u"",assetClass=u"E",ticker=x,cnSpell=u"",field=u"",pandas="1")['secShortName'])
df
```

```python
df['date'] = pd.to_datetime(df['date'])
df['date'] = df['date'].dt.strftime('%Y%m%d')
# 求问：下面这段代码写的跟shit一样，有没有人能告诉我如果要利用两个参数的时候，apply无能的时候，我该怎么用比较pandas的方法来写？
for i in range(len(df.index)):
    date = df['date'].iloc[i]
    secID = df['secID'].iloc[i]
    histdata = DataAPI.MktEqudGet(tradeDate=date,secID='',ticker=secID,beginDate=u"",endDate=u"",isOpen="",field=u"",pandas="1")
    df.loc[df['secID']==secID,'market_price'] = histdata['closePrice'][0]*histdata['accumAdjFactor'][0]
df['ratio'] = df['subscription'] / df['market_price']
total_cost = df['cost'].sum()
df['weight'] = df['cost'] /total_cost
df
```

```python
discount = (df['ratio'] * df['weight']).sum()
discount
```

# 总结：大体而言，八折拿股票

定增出现相较于股票的折价这是最正常不过的，试想，如果定增价格相当于市场价或者更高，那么有谁会来参与定增，二级市场购买不是一个更好的主意么？

定增的折价 = 股票的流动性补偿

假设我们认为有效市场的假设成立，那么定增股价相较于流通股的折价，就是流动性补偿。首先，它应该高于无风险利率，假设是5.5%；其次，之所以是高于是因为股票相较于无风险资产，应该再加上一个风险补偿。在此的基础上，如果市场不是完全有效的，那么我们再应该根据市场情况重新评估择时和择股的影响。
