# 

    估计问题可以粗略分为如下两种：总体分布已知，总体分布未知。
    若已知总体分布为正太分布或者指数分布时，可以对总体参数进行严格估计。本文档集中讨论总体分布为正太分布或指数分布的情形。

一.正太分布
正太分布具有如下几个重要性质：
1. 若X1，X2...Xn~iid N（u，sigma**2），则这n个随机变量的平均值服从正太分布N（u，sigma**2/n），而Z=（(X1+X2+..Xn)/n-u）/(sigma/sqrt(n))~N(0,1), 即服从标准正太分布

温馨提示：注意区分这条性质和中心极限定理的区别，此性质仅适用于独立同正太分布的随机变量序列，但对于任意的n都成立。可以改变下列代码中的n值来观察这一现象。

由于变量Z中同时含有总体分布的平均值和方差，因此仅适用于方差已知的情况，用来估计平均值，对应的假设检测即为Z检测


```python
import numpy as np
import scipy.stats as sps 
import matplotlib.pyplot as plt
n=8
loc=4
scale=2
Xnm=sps.norm.rvs(loc,scale,size=(n,100000))
Xm=np.mean(Xnm,axis=0)
Z=(Xm-loc)/(scale/np.sqrt(n))
x=np.linspace(start=-5, stop=5,num=100)

plt.hist(Z,bins=100,normed=True,color='b')
plt.plot(x,sps.norm.pdf(x),'r')
plt.legend(['sum_norm','norm'])
plt.xlabel('x')
plt.ylabel('pdf')
```

2.若X1，X2...Xn~iid N（u，sigma**2），则 ZZ=(对i求和_（（Xi-X平均）**2）)/sigma**2 ~卡方（n-1）

同上，此性质仅适用于独立同正太分布的随机变量序列，但对于任意的n都成立

由于此ZZ变量（笔者自己命名）中仅含有总体方差，因此适合对总体方差进行估计

```python
n=7
loc=4
scale=2
Xnm=sps.norm.rvs(loc,scale,size=(n,100000))
Xm=np.mean(Xnm,axis=0)
ZZ=np.sum((Xnm-Xm)**2,axis=0)/scale**2
x=np.linspace(start=0, stop=30,num=100)
plt.hist(ZZ,bins=100,normed=True,color='b')
plt.plot(x,sps.chi2.pdf(x,df=n-1),'r')
plt.legend(['sum_norm2','chi2'])
plt.xlabel('x')
plt.ylabel('pdf')
plt.show()
```


3.Z变量和ZZ变量相互独立的。
前面我们提过，若A为正太随机变量，B为卡方(n)随机变量，且两者相互独立，则T=A/sqrt（B/n）服从t(n)分布结合性质1，2，3，我们可以构造如下随机变量T
T=Z/sqrt（ZZ/(n-1)），易知 T~t(n-1)
值得注意的是，T变量中只包含总体平均值，而不包含总体方差，因此适用于总体均值和方差都未知的情况，用来对总体均值进行估计。对应的建设检验即为T检验

```python
n=9
loc=4
scale=2
Xnm=sps.norm.rvs(loc,scale,size=(n,100000))
Xm=np.mean(Xnm,axis=0)
Z=(Xm-loc)/(scale/np.sqrt(n))
ZZ=np.sum((Xnm-Xm)**2,axis=0)/scale**2
T=Z/np.sqrt(ZZ/(n-1))
x=np.linspace(start=-5, stop=5,num=100)
plt.hist(T,bins=100,normed=True,color='b')
plt.plot(x,sps.t.pdf(x,df=n-1),'r')
plt.legend(['Z/sqrt(ZZ/(n-1))','t'])
plt.xlabel('x')
plt.ylabel('pdf')
```

```python
二.指数分布
   指数分布具有如下性质，若X1，X2...Xn~iid Expon（lamada,x）=lamada*exp(-lamada*x)，则H=2*lamada*（X1+X2+..Xn）服从卡方（2n）分布，利用此性质可以对指数分布的lamada进行估计。
    值得注意的是，python中指数分布定义为Expon=1/scale*exp（-（x-loc）/scale），即scale为指数分布平均值，亦即1/lamada
```

```python
n=4
lamada=5
Xnm=sps.expon.rvs(loc=0,scale=1.0/lamada,size=(n,100000))
Xm=np.sum(Xnm,axis=0)
H=2*lamada*Xm
x=np.linspace(start=0, stop=50,num=100)
plt.hist(H,bins=100,normed=True,color='b')
plt.plot(x,sps.chi2.pdf(x,df=2*n),'r')
plt.legend(['sum_expon','chi2'])
plt.xlabel('x')
plt.ylabel('pdf')
```
