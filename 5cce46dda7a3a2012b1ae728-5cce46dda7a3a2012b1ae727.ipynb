```python
# built-in
import time
import operator
import datetime as dt
import collections
from functools import wraps

# 3-party
import pandas as pd
pd.options.display.max_columns = 100
pd.options.display.max_rows = 300

import numpy as np

import seaborn
from matplotlib import pylab
import matplotlib.pyplot as plt
from matplotlib.pyplot import *

from CAL.PyCAL import *
font.set_size(22)

# personal lib
from lib import report
```

```python

def bei_shang_week_industry():
    "北上资金持股占比变动"
    stats_Freq=[u'近一周']
    data = DataAPI.HKshszShareChgGet(secID=set_universe('A'),statsFreq=stats_Freq ,field=['ticker', 'statsFreq','changePct'])
    industry_ticker = DataAPI.EquIndustryGet(secID=set_universe('A'),
                                             field=['ticker', 'secID', 'secShortName', 'industryName1','isNew', 'industry'])
    industry_ticker = industry_ticker[(industry_ticker.isNew == 1) & (industry_ticker.industry == '申万行业分类')]
    data = data.merge(industry_ticker[['ticker', 'industryName1']], on=['ticker'])
    industry_data = data.groupby('industryName1').sum()
    #取changePct，降序，替换原来df（inplace）
    industry_data.sort(columns=['changePct'], ascending=True, inplace=True)
    ### plot
    tmp1 = industry_data[:5]
    tmp2 = industry_data[-10:]
    tmp = tmp1.append(tmp2)
    try:
        ax = tmp[::-1].plot(y=['changePct'], kind='bar', figsize=(14, 7), color='r', title=u'申万一级行业 本周 持股百分比变化')
        for label in ax.get_xticklabels(): 
            label.set_fontproperties(font) 
        report.fig_style(ax, [u'北上资金持股占比变动'], 'best', u'行业名称')
    except:
        print 'data not ready'
    return
bei_shang_week_industry()

def bei_shang_mounth_industry():
    "北上资金持股占比变动"
    stats_Freq=[u'近一月']
    data = DataAPI.HKshszShareChgGet(secID=set_universe('A'),statsFreq=stats_Freq ,field=['ticker', 'statsFreq','changePct'])
    industry_ticker = DataAPI.EquIndustryGet(secID=set_universe('A'),
                                             field=['ticker', 'secID', 'secShortName', 'industryName1','isNew', 'industry'])
    industry_ticker = industry_ticker[(industry_ticker.isNew == 1) & (industry_ticker.industry == '申万行业分类')]
    data = data.merge(industry_ticker[['ticker', 'industryName1']], on=['ticker'])
    industry_data = data.groupby('industryName1').sum()
    #取changePct，降序，替换原来df（inplace）
    industry_data.sort(columns=['changePct'], ascending=True, inplace=True)
    ### plot
    tmp1 = industry_data[:5]
    tmp2 = industry_data[-10:]
    tmp = tmp1.append(tmp2)
    try:
        ax = tmp[::-1].plot(y=['changePct'], kind='bar', figsize=(14, 7), color='r', title=u'申万一级行业 近一月 持股百分比变化')
        for label in ax.get_xticklabels(): 
            label.set_fontproperties(font) 
        report.fig_style(ax, [u'北上资金持股占比变动'], 'best', u'行业名称')
    except:
        print 'data not ready'
    return
bei_shang_mounth_industry()

```
