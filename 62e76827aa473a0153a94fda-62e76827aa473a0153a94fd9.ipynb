```python
def ETFStockUniverse(etf_universe):
    u_list = []
    for idx in etf_universe:
        df_1 = DataAPI.FundHoldingsGet(secID=idx,ticker="",reportDate=u"",beginDate=u"",endDate=u"",
               secType="E",holdingTicker="",holdingSecID="",field=u"holdingSecID,publishDate",pandas="1")
        u_list.extend(list(df_1["holdingSecID"]))
    u_list_r = list(set(u_list))
    return u_list_r
```
