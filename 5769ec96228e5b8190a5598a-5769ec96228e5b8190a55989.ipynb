```python
import pandas as pd
import seaborn
from CAL.PyCAL import font

# 准备统计的四只股票
tickers = [('02202','000002', u'万科'),
           ('03968','600036', u'招商银行'), 
           ('00489','600881', u'东风集团'), 
           ('02628','601628', u'中国人寿'),  
          ]

begin_date = '20150101'
HSAHP_df = pd.DataFrame()

# HKD:CNY
# 港币：人民币 
hdf1 = DataAPI.ChinaDataExchangeRateGet(indicID=u"M110000003",beginDate=begin_date,field=u"periodDate,dataValue",pandas="1")
hdf2 = DataAPI.HKDataExchangeRateGet(indicID=u"G140000102",beginDate=begin_date,field=u"periodDate,dataValue",pandas="1")

hdf = pd.merge(hdf1,hdf2,on=['periodDate'])
hdf['huilv'] = hdf['dataValue_x']/hdf['dataValue_y']
hdf['tradeDate'] = hdf['periodDate']

for h_ticker,a_ticker,name in tickers:
    df = DataAPI.MktHKEqudGet(ticker=h_ticker, beginDate=begin_date, field='tradeDate,closePrice')
    df2 = DataAPI.MktEqudGet(ticker=a_ticker, beginDate=begin_date, field='tradeDate,closePrice')
    tmp = pd.merge(df,df2,on=['tradeDate'])
    ndf = pd.merge(hdf,tmp,on=['tradeDate'])
    ndf.sort('tradeDate',inplace=True, ascending=True)
    ndf.reset_index(inplace=True)
    HSAHP_df[name] =ndf['closePrice_y']/ndf['closePrice_x']/ndf['huilv']
    HSAHP_df['tradeDate'] = ndf['tradeDate']
```

首先看一下万科AH比价曲线，销魂的`一只穿云箭`走势，忍不住用最华丽的语言来形容。

```python
y = [u'万科']
ax = HSAHP_df.plot(x='tradeDate', y=y, figsize=(12, 6), title=u'AH比价走势', linewidth='3')
ax.legend(y, prop=font, loc='upper right')
ax.title.set_font_properties(font)
```
