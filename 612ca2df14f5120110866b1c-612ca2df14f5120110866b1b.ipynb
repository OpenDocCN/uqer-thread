# 牛熊指标（换手率与波动率）

# 原理
来自《华泰金工量化择时系列：牛熊指标在择时轮动中的应用探讨》，使用牛熊指标双均线进行择时。
# 构建过程
1. 牛熊指数
	1. 指数换手率与波动率
		- 波动率：收益率标准差
		- 换手率：指数成分股成交量合计/成分股流通股份数合计
	1. 构建牛熊指数
		牛熊指标=N日换手率均值/N日波动率
2. 择时
    使用30日牛熊指标（短线）和250日牛熊指标（长线），短线位于长线上方持仓，短线位于长线下方清仓。

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
```

```python
def get_working_days(start_date,end_date):
    from CAL.PyCAL import *
    cal = Calendar('CHINA.SSE')
    day_list=cal.bizDatesList(start_date,end_date)
    str_day_list=[]
    for d in day_list:
        str_day_list.append(d.strftime('%Y%m%d'))
    return str_day_list
```

# 定义参数。想测试其他指数和时间段在这里修改。

```python
index_ticker='000016'
index='000016.ZICN'
beginDate='20130101'
endDate='20210827'
```

# 计算每日指标函数，包括当日收益率和换手率。

```python
#计算某日的三个指标
def cal_ind(index,tradeDate):
    #获取当日收益计算波动率
    price=DataAPI.MktIdxdGet(ticker=index,tradeDate=tradeDate,field=u"tradeDate,turnoverVol,CHGPct",pandas="1")
    price.set_index('tradeDate',inplace=True)
    #取得当天所有成分股的流通股份数
    stocks=DataAPI.IdxConsGet(ticker=index,intoDate=tradeDate,field=u"consTickerSymbol",pandas="1")['consTickerSymbol'].tolist()
    stock_price=DataAPI.MktEqudGet(ticker=stocks,tradeDate=tradeDate,field=u"ticker,closePrice,negMarketValue",pandas="1")
    stock_price['amount']=stock_price['negMarketValue']/stock_price['closePrice']
    amount=stock_price['amount'].sum()
    #计算换手率
    turnover=price.iloc[0]['turnoverVol']/amount
    chgPct=price.iloc[0]['CHGPct']
    return turnover,chgPct
```

# 计算每日基础指标（换手率和收益率），合成Dataframe

```python
cal=get_working_days(beginDate,endDate)
result={}
for date in cal:
    tmp={}
    tmp['turnover'],tmp['chgpct']=cal_ind(index_ticker,date)
    result[date]=tmp
data=pd.DataFrame(result).T
```

# 加入每日指数点位便于对比

```python
index_data=DataAPI.MktIdxdGet(beginDate=beginDate,endDate=endDate,indexID=index,field=u"tradeDate,closeIndex",pandas="1")
index_data['tradeDate']=index_data['tradeDate'].str.replace('-','')
index_data.set_index('tradeDate',inplace=True)
all_data=pd.merge(data,index_data,left_index=True,right_index=True)
```

# 计算每日牛熊指数

```python
N=30#短线参数
M=250#长线参数
all_data['chgPct_std_long']=all_data['chgpct'].rolling(window=M).std()#长线收益率标准差
all_data['turnover_mean_long']=all_data['turnover'].rolling(window=M).mean()#长线换手率均值
all_data['chgPct_std_short']=all_data['chgpct'].rolling(window=N).std()#短线收益率标准差
all_data['turnover_mean_short']=all_data['turnover'].rolling(window=N).mean()#短线换手率均值
all_data['bull_bear_long']=all_data['turnover_mean_long']/all_data['chgPct_std_long']#长线牛熊指标
all_data['bull_bear_short']=all_data['turnover_mean_short']/all_data['chgPct_std_short']#短线牛熊指标
#以下构建买卖信号，短线大于长线为指数点位，否则为0（便于绘图）
all_data['signal']=(all_data['bull_bear_short']-all_data['bull_bear_long'])
all_data['signal']=((all_data['signal']+abs(all_data['signal']))/2)/all_data['signal']*all_data['closeIndex']
all_data.index=pd.DatetimeIndex(all_data.index)           
```

# 绘制结果。
绿线是指数点位（右轴），红色是短线，蓝色是长线。直观看能避开熊市，但是信号比较频繁。

```python
# 交叉点对比
plt.Figure(figsize=(50,50))
fig, ax1 = plt.subplots(figsize=(20, 10))
ax2=ax1.twinx()
ax1.plot(all_data.index,all_data['bull_bear_short'],'r-')
ax1.plot(all_data.index,all_data['bull_bear_long'],'b-')
ax2.plot(all_data.index,all_data['closeIndex'],'g-')
plt.show()
```

# 将买卖信号附加到指数上，便于观察持仓时机。

```python
# 绘制图形
plt.Figure(figsize=(50,50))
fig, ax1 = plt.subplots(figsize=(20, 10))
ax1.plot(all_data.index,all_data['closeIndex'],color='gray')
ax1.scatter(all_data.index,all_data['signal'],c='#FF0000')
plt.show()
```

# 测试结果分析
1. 优点
- 能避开熊市，抓住大部分牛市。
- 没有复杂的数学模型（只使用了标准差和均值），便于理解。
2. 缺点
- 发出信号较晚，无法避开熊市开头的大跌。（比如15年6月的大跌7月才发出信号，对于这种急跌的无效。）
- 假信号多。
