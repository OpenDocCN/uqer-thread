# DataFrame的一些基本操作

# DataFrame的创建

```python
df = pd.DataFrame({'one':[1,2,3],'two':[4,5,6]})
df1 = pd.DataFrame({'one':[1,2,3],'two':[4,5,6]},index=['a','b','c'])
print df1

print '-'*11

df2 = pd.DataFrame({'one':pd.Series([1,2,3],index=['a','b','c']),'two':pd.Series([1,2,3,4],index=['b','a','c','d'])})
print df2
```

# DataFrame常用属性：
 index，T，columns，values，describe()

```python
print df2
# 行索引
print df2.index
# 值
print df2.values
# 列索引
print df2.columns
# 描述性统计
df2.describe()
```

# DataFrame索引和切片

```python
print df2
# 只能先列后行，可以查看df2['one']，但是不能df2['a'],切行只能df2.loc['a',:]
print df2['one']['a']
print df2.loc[['a','c'],'one']
```

# DataFrame数据对齐与缺失处理

```python
df3 = pd.DataFrame({'two':[1,2,3,4],'one':[4,5,6,7]},index=['c','d','b','a'])
print df2
print '-'*12
print df3
print '-'*12
print df2.fillna(0)
print '-'*12
print df2.dropna(how='any')
print '-'*12
print df2.dropna(how='all')
print '-'*12
print df2.dropna(axis=1)
```

# pandas其他常用方法:
- mean(axis=0,skipna=False)          对列（行）求平均数
- sum(axis=1)                        对列（行）求和
- sort_index(axis,...,ascending)     对列（行）索引排序
- sort_values(by,axis,ascending)     按某一列（行）的值排序
- Numpy的通用函数适用于pandas

```python
print df2
print '-'*12
print df2.mean()
print '-'*12
print df2.mean(axis=1)
print '-'*12
print df2.sort_values(by='two',ascending=False)
print '-'*12
# 有缺失值的情况nan排序统一放在最后
print df2.sort_values(by='two',ascending=False)
print '-'*12
print df2.sort_index(ascending=False,axis=1)
```

# 时间对象处理
## 时间序列类型：
·时间戳：特定时刻
·固定时期：2017年7月
·时间间隔：起始时间-结束时间
## python标准库：datetime
## 灵活处理时间对象：dateutil.parser.parse()
## 成组处理事件对象：pd.to_datetime()

```python
import pandas as pd
import datetime as dt
import dateutil as dtu

# 第二个datetime是dt模块下的类
print dt.datetime.strptime('2010-01-01','%Y-%m-%d')
# 我们发现上面这种写法还得规定格式，很麻烦，所以推荐使用dateutil
print dtu.parser.parse('2010-01-01')
print dtu.parser.parse('2010/01/01')
print dtu.parser.parse('20100101')

print pd.to_datetime(['2010-01-01','2010-JAN-02'])
```

# 时间对象生成

```python
print pd.date_range('2010-01-01','2010-01-06')
print pd.date_range('2010-01-01',periods=6)
print pd.date_range('2010-01-01',periods=30,freq='W-MON')
print pd.date_range('2010-01-01',periods=30,freq='W-MON')[0].to_pydatetime()
```
