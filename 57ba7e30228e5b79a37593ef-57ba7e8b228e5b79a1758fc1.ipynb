>枢轴点（Pivot Point）是一套非常“单纯”的阻力支持体系，大概是10年前一个做期货的高手发明的方法，至今已经广泛的用在股票、期货、国债、指数等高成交量的商品上。经典的Pivot Point是7点系统，就是7个价格组成的，目前广泛使用的13点系统，其实都是一样的，不过是多加了6个价格罢了，用于大成交量的商品。

具体的计算公式如下：

P= ( H + L + 2C ) / 4 {H代表高价位, L代表低价位, C代表收市价}

这个计算出的P值，是当时的市场绝对均价

下文用到P值公式是变体。

下面的就是原理公式：

pivot:= (high + low + close) / 3;（用前一天的最高、最低和收盘）

r1:= 2*pivot - low;

s1:= 2*pivot - high;

r2:= pivot + (r1-s1);

s2:= pivot - (r1-s1);

r3:= high + 2 * (pivot-low);

s3:= low - 2 * (high - pivot);

r是压力位，s是支撑位。我先在文华上做了一个指标，先用肉眼观察一下是怎样的。

`买入条件`

1、最近10天的跌幅没有大于5%

2、最近30个交易日没有停盘过

3、如果s3ma10ma10>s3ma20ma10(也就是s3的ma10的ma10和s3的ma20的ma10)这个是不是有点奇葩，完全是肉眼用归纳法归纳出来的，类似于均线多头

5、前一个交易日的收盘价大于s3ma20ma10，跌破这个位置，基本就是大跌了

6、s3ma10>s3ma10ma10,

7、s3ma20ma10相比再前一个交易日是上涨的

`卖出条件`

1、止损3%

2、前一个交易日最高价大于（r3ma10-s3ma10）/10*8+s3ma10

3、买入后每天平均涨幅低于1%

```python
from CAL.PyCAL import *
import numpy as np
import talib.func as tl
import lib.pivot as pivot


start = '2013-01-01'                       # 回测起始时间
end = '2016-07-31'                         # 回测结束时间
benchmark = 'HS300'                        # 策略参考标准
universe =set_universe('A') # 证券池，支持股票和基金
capital_base = 1000000                      # 起始资金
freq = 'd'                                 # 策略类型，'d'表示日间策略使用日线回测，'m'表示日内策略使用分钟线回测
refresh_rate = 1                           # 调仓频率，表示执行handle_data的时间间隔，若freq = 'd'时间间隔的单位为交易日，若freq = 'm'时间间隔为分钟
cal=Calendar('China.SSE')



#DataAPI.settings.cache_enabled = True

def initialize(account):                   # 初始化虚拟账户状态
    account.mairushijian={}

def handle_data(account):                  # 每个交易日的买入卖出指令
    buylist=[]
    sellist=[]
    last30day=cal.advanceDate(account.previous_date,'-30B')
    STlist = DataAPI.SecSTGet(secID=universe, beginDate=account.previous_date, endDate=account.previous_date, field=['secID'],pandas='1')['secID'].tolist()
    #ST列表
    stocklist=[s for s in account.universe if s not in STlist]
   
    for stk in stocklist :
        if stk not in account.valid_secpos:

            prices=DataAPI.MktEqudAdjGet(secID=stk,beginDate=last30day,endDate=account.previous_date,field=u"secID,tradeDate,openPrice,highestPrice,lowestPrice,closePrice",pandas="1").dropna()#获取开盘收盘价用来计算PP
            if len(prices['closePrice'])<31 or (0.00 in prices['openPrice'].tolist()):
                continue
       
            prices['zhangdie']=(prices['closePrice']-prices['closePrice'].shift(1))/prices['closePrice'].shift(1)
            if min(prices['zhangdie'][20:].tolist())<-0.05:#如果最近10天有跌幅超过5%的，该只股票就不考虑
                continue
            prices=pivot.pivot(prices)     #调用自定义函数，计算枢轴点系统的支撑位置，以及一些的简单的参数变换
            
            
            if prices['s3ma10ma10']>prices['s3ma20ma10']  and prices['closePrice']<=((prices['r3ma10']-prices['s3ma10'])/10+prices['s3ma10']) and prices['closePrice']>prices['s3ma20ma10'] and prices['s3ma10']>prices['s3ma10ma10'] and prices['s3ma10ma10qushi']>1:

                buylist.append(stk)

    for stock in account.valid_secpos.keys():
        prices=DataAPI.MktEqudAdjGet(secID=stk,beginDate=last30day,endDate=account.previous_date,field=u"secID,tradeDate,openPrice,highestPrice,lowestPrice,closePrice",pandas="1").dropna()#获取开盘收盘价用来计算PP
        if len(prices['closePrice'])<31 or (0.00 in prices['openPrice'].tolist()):
            continue
        
        prices=pivot.pivot(prices)  #调用自定义函数，计算枢轴点系统的支撑位置，以及一些的简单的参数变换
        
        price=DataAPI.MktEqudAdjGet(beginDate=account.mairushijian[stock],endDate=account.previous_date,secID=stock,field=u"highestPrice",pandas="1")
        maxprice=max(price['highestPrice'].tolist())
        
        
        if account.referencePrice[stock]<account.valid_seccost[stock]*0.97  or maxprice>=(prices['r3ma10']-prices['s3ma10'])/10*8+prices['s3ma10']  or (prices['closePrice']<account.valid_seccost[stock]*(1.0+0.01* len(cal.bizDatesList(account.mairushijian[stock],account.current_date)))):
            sellist.append(stock)
        else:
            buylist.append(stock)
    for stk in sellist:
        order_to(stk,0)
    prices = account.referencePrice
    for stk in buylist:
        if stk not in account.valid_secpos.keys() and stk in account.universe:
            account.mairushijian[stk]=account.current_date
        if stk in account.universe:
            minpct=min(0.1,0.98/len(buylist))
            order_pct_to(stk,0.98/len(buylist))
```

下面是lib.pivot的自定义内容，因为这个回测不是很成功，所以为了突出主要内容，我把这部分代码折叠了。

```python
#已折叠
def pivot(prices):
    import talib.func as tl
    prices['pp']=(prices['highestPrice']+ prices['lowestPrice'] +2*prices['closePrice'])/4
    prices['r3']=prices['highestPrice']-(2*(prices['lowestPrice']-prices['pp']))
    prices['s3']=prices['lowestPrice']-(2*(prices['highestPrice']-prices['pp']))
    prices['r3ma10']=tl.MA(prices['r3'].values,timeperiod=10)
    prices['s3ma20']=tl.MA(prices['s3'].values,timeperiod=20)
    prices['s3ma10']=tl.MA(prices['s3'].values,timeperiod=10)
    prices['s3ma10ma10']=tl.MA(prices['s3ma10'].values,timeperiod=10)
    prices['s3ma20ma10']=tl.MA(prices['s3ma20'].values,timeperiod=10)
    
            
    prices['pp']=prices['pp'].shift(1)
    prices['r3']=prices['r3'].shift(1)
    prices['s3']=prices['s3'].shift(1)
    prices['r3ma10']=prices['r3ma10'].shift(1)
    prices['s3ma20']=prices['s3ma20'].shift(1)
    prices['s3ma10']=prices['s3ma10'].shift(1)
    prices['s3ma10ma10']=prices['s3ma10ma10'].shift(1)
    prices['s3ma20ma10']=prices['s3ma20ma10'].shift(1)
    prices['s3ma20ma10qushi']=prices['s3ma20ma10']/prices['s3ma20ma10'].shift(1)
    prices['s3ma10ma10qushi']=prices['s3ma10ma10']/prices['s3ma10ma10'].shift(1)
            
            
    prices=dict(prices.iloc[-1,:])
    return prices
```
