# Ta-lib 使用心得

```python
import numpy as np
import pandas as pd
import talib as ta
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.patches as patches
filed = ['ticker','tradeDate','openPrice','highestPrice','lowestPrice','closePrice','turnoverVol']
data = DataAPI.MktMFutdGet(tradeDate=u"",mainCon=u"",contractMark=u"",contractObject=u"IF",startDate=u"20170201",endDate=u"20170225",field=filed,pandas="1")
data = data.set_index(data.tradeDate)
del data['tradeDate']
```

```python
def mkt_plot(quotes, sec):           # quotes:行情数据-Dateframe类型    sec：标题    
    __color_balck__, __color_green__, __color_yellow__, __color_purple__,__linewidth__ = '#0F0F0F', '#00FFFF', '#EE9A00', '#9900CC' , 2
    fig = plt.figure(figsize=(11,5))
    ax1 = fig.add_axes([0, 1, 1, 1])               
    ax1.set_title(sec,  fontsize=15)
    ax1.grid(True, axis='y')
    ax1.set_xlim(-1, len(quotes)+1)
    for i in range(len(quotes)):
        close_price,open_price = quotes.ix[i, 'closePrice'], quotes.ix[i, 'openPrice']
        high_price, low_price = quotes.ix[i, 'highestPrice'], quotes.ix[i, 'lowestPrice']
        trade_date = quotes.index[i]
        if close_price > open_price:#画阳线
            ax1.add_patch(patches.Rectangle((i-0.2, open_price), 0.4, close_price-open_price, fill=False, color='r'))
            ax1.plot([i, i], [low_price, open_price], 'r')
            ax1.plot([i, i], [close_price, high_price], 'r')
        else:#画阴线
            ax1.add_patch(patches.Rectangle((i-0.2, open_price), 0.4, close_price-open_price, color='g'))
            ax1.plot([i, i], [low_price, high_price], color='g')
    ax1.set_title("Price", fontsize=15, loc='left', color='r')
    #设置x轴标签
    ax1.set_xticks(range(0,len(quotes),5))#位置
    ax1.set_xticklabels([quotes.index[i] for i in ax1.get_xticks()] , rotation=20)#标签内容
    op, high, low, close  = np.array(quotes['openPrice']), np.array(quotes['highestPrice']),np.array(quotes['lowestPrice']),np.array(quotes['closePrice'])
    CDLDOJI = ta.CDLDOJI(op, high, low, close)                                       #十字
    CDLRICKSHAWMAN = ta.CDLRICKSHAWMAN(op, high, low, close)                        #黄包车夫
    ax2 = ax1.twinx()
    ax2.plot(CDLDOJI, marker='+',  color=__color_yellow__,  label='CDLDOJI')
    ax2.plot(CDLRICKSHAWMAN ,marker='*',color=__color_purple__,  label='CDLRICKSHAWMAN ')
    #图例
    ax2.legend(loc='Upper Left')
    return fig
```

```python
mkt_plot(data,'IFL0')
#下图是IF主力合约的K线图。
#如图：所以紫色的为识别了黄包车夫模式的K线，黄色的是十字K线。
```
