```python
import pandas as pd 
import numpy as np

myBeginDate0 = '2017-01-17'
myEndDate0 = '2017-03-22'

myBeginDate1 = myBeginDate0.replace('-','')
myEndDate1 =  myEndDate0.replace('-','')

# *************001 返回 基金 代码和名称*****************************************************************
#etf0 = DataAPI.FundGet(etfLof=u"ETF",secID=u"",ticker=u"",category=u"E",idxID=u"",idxTicker=u"",listStatusCd=u"L",operationMode=u"O",field=u"",pandas="1")

etf0 = DataAPI.FundGet(etfLof=u"",secID=u"",ticker=u"",category=u"",idxID=u"",idxTicker=u"",listStatusCd=u"L",operationMode=u"O",field=u"",pandas="1")


etf0 =  etf0[(etf0.delistDate.isnull())]
etf0 = etf0[['secID','ticker','secShortName','tradeAbbrName','idxTicker','idxShortName']]
#etf0


price1 = DataAPI.MktFunddAdjBfGet(secID=etf0.secID,ticker=u"",beginDate=myBeginDate1,endDate=myBeginDate1,field=u"secID,ticker,secShortName,tradeDate,closePrice",pandas="1")
#secID,ticker,secShortName,tradeDate,closePrice
#price1

price2= DataAPI.MktFunddAdjBfGet(secID=etf0.secID,ticker=u"",beginDate=myEndDate1,endDate=myEndDate1,field=u"secID,tradeDate,closePrice",pandas="1")

#price2

price3 = pd.merge(price1,price2,how='inner',left_on='secID',right_on='secID')
#price3

#计算变化幅度
price3['CHGPct'] =100*(price3['closePrice_y'] - price3['closePrice_x'])/price3['closePrice_x'] 
price3 = price3.sort(columns=['CHGPct'], ascending=[False])
price3 = price3[['ticker','secShortName','CHGPct']]

price3 = price3[price3['CHGPct']>0]
price3 = price3[price3['CHGPct'] > price3['CHGPct'].mean()]

print '开始日期：'+myBeginDate1+'　结束日期：'+myEndDate1
price3
#price3.head(50)
```
