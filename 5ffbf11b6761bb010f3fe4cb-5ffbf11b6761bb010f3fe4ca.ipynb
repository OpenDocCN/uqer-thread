```python
import pandas as pd
import numpy as np
import datetime, time
import matplotlib.pyplot as plt
```

```python
# 选取月度频率计算收益率分化程度，取月底日起
df_trading_cal = DataAPI.TradeCalGet(exchangeCD=u"XSHG,XSHE",beginDate=u"20000101", endDate=datetime.datetime.today())
ME_dates = list(df_trading_cal[df_trading_cal['isMonthEnd']==1]['calendarDate'].unique())
ME_dates[-5:]
```

```python
# 汇总各月各股票的交易信息统计 - 月度收益率 return，平均换手率 avgTurnoverRate，近两年收益率标准差 sdReturn24
tic = time.time()
df_p_total = pd.DataFrame()
for eval_dt in ME_dates:
    df_p = DataAPI.MktEqumAdjGet(secID='', monthEndDate=eval_dt, isOpen=1, field=['secID','endDate','return','avgTurnoverRate','sdReturn24'])
    df_p_total = pd.concat([df_p_total, df_p])
    # display progress
    if eval_dt[5:7] == '12': print('Pulled metrics for {:s} - using {:.2f} min in total'.format(eval_dt, (time.time()-tic)/60))

df_p_total.head()
```

```python
# 添加行业分类信息 - 用最新的申万一级行业
sw_map_frame = DataAPI.EquIndustryGet(industryVersionCD=u"010303", field=['secID','intoDate', 'outDate', 'industryName1', 'industryID1','isNew'])
sw_map_frame = sw_map_frame.sort_values(by=['secID','intoDate'], ascending=[True,False]).drop_duplicates(subset=['secID'], keep='first')[['secID','industryName1','industryID1']] # to include delisted stocks
df_full = df_p_total.merge(sw_map_frame, how='left', on='secID')
df_full = df_full[df_full['sdReturn24'] > 0]
df_full.head()
```

```python
# 观察 月度收益率 与 平均换手率 在月度横截面数据上的标准差，作为分化程度的指标
for metric in ['return','avgTurnoverRate']:
    df_full.groupby('endDate')[[metric]].std().plot()
```

可以看到收益率的分化程度在牛熊转换时快速上升，最近一次高点在2015/05月出现；而换手率的分化自2002年后一直上升，从15年熊市之后便一直保持高位。

```python
# 详细的指标请参考下表
df_summary_market = df_full.groupby('endDate')[['return','avgTurnoverRate']].std()
df_summary_market.loc['2014':'2016']
```

### 不同行业的收益率、换手率分化程度如何？- 下面检验最新时点上的行业分化程度

```python
# 将最近一月的各股数据按行业分类，然后按收益率的标准差进行降序排列
recent_date = df_full['endDate'].max()
print('Most recent date: {:s}'.format(recent_date))
recent_snapshot = df_full[df_full['endDate']==recent_date].pivot_table(index='industryName1', aggfunc=('mean','std'))
recent_snapshot.sort_values(by=('return','std'), ascending=False)
```

在2020/12月，食品饮料、电气设备、采掘是收益率分化最大的三个行业

```python
# 食品饮料
df_full[df_full['industryName1']=='食品饮料'].groupby('endDate')[['return']].std().plot()
```

```python
# 电气设备
df_full[df_full['industryName1']=='电气设备'].groupby('endDate')[['return']].std().plot()
```

```python
# 采掘
df_full[df_full['industryName1']=='采掘'].groupby('endDate')[['return']].std().plot()
```

通过对分化程度前三大的行业（食品饮料、电气设备、采掘）进行纵向观察，发现这三个行业（尤其是食品饮料）的收益率分散程度都在近期有明显上升。


```python
# 完整的数据如下表
df_full_summary = df_full.pivot_table(index=['industryName1','endDate'], aggfunc=('mean','std')).reset_index()
df_full_summary.head()
```

### 对量化选股价值的探讨

通过量化模型筛选股票，旨在用规则选出表现较优的股票，所以收益率分散程度可以描述量化选股能提供价值的上限。如果收益率分散程度很小，则量化模型较难提供样本内的超额收益。
因此下一步简单计算各行业在近期的分化程度。

```python
df_ind_dispersion = df_full_summary[df_full_summary['endDate']>='2019'].groupby('industryName1')[[('return','std')]].mean().reset_index()
df_ind_dispersion.sort_values(by=[('return','std')], ascending=False)
```
