## 获取高转送数据
---
这里我们调用DataAPI，EquDivGet获取历史上各只股票的分红送转情况，参考了兴业证券的研报，将高转送定义为转股比例加上送股比例大于0.5。

```python
"""
加载包
"""
import pandas as pd
import numpy as np
import datetime
import matplotlib.pylab as plt
from CAL.PyCAL import *
import statsmodels.api as sm 
from sklearn.linear_model import LogisticRegression
cal = Calendar('China.SSE')
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.learning_curve import learning_curve
```

```python
def get_universe(trade_date):
    """
    获取动态股票列表，根据交易日获取所有上市的股票ID
    """
    eq_msg = DataAPI.EquGet(equTypeCD=u"A",secID=u"",ticker=u"",listStatusCD=u"L,S,DE,UN",
               field=u"secID,listDate,delistDate",pandas="1")
    eq_msg = eq_msg.fillna(datetime.datetime.today().strftime('%Y-%m-%d'))
    eq_msg = eq_msg[(eq_msg['listDate']<trade_date) & (eq_msg['delistDate']>trade_date)]
    return eq_msg['secID'].tolist()
```

```python
"""
获取自A股开市以来所有的股票ID（包含已退市的股票）
"""
universe = DataAPI.EquGet(equTypeCD=u"A",secID=u"",ticker=u"",listStatusCD=u"L,S,DE,UN",field=u"secID",pandas="1")['secID'].tolist()#包括退市的股票
```

```python
"""
获取股票的分红信息，包含股票ID，公布日期，分红截止年度，每股转增股比例和每股送股比例
EquDivGet获取股票分红信息
"""
div_summary = DataAPI.EquDivGet(exDivDate="",eventProcessCD=u"",secID=universe,ticker=u"",
                  beginDate=u"2011-01-01",endDate=u"",field=u"secID,publishDate,endDate,perShareDivRatio,perShareTransRatio",pandas="1")
div_summary = div_summary.dropna(subset=['publishDate']) #公布日期为空的数据清洗掉
div_summary = div_summary.fillna(0) #其他数据为空填充0
div_summary['TotalRatio'] = div_summary['perShareDivRatio'] + div_summary['perShareTransRatio'] #转股和送股比例和
high_amount = div_summary[div_summary['TotalRatio'] > 0.5].reset_index(drop=True) #筛选站送股比例大于0.5的股票
high_amount['year'] = high_amount['publishDate'].map(lambda x:x[0:4]) #获取送转股年度
```

## 建立模型
---
目标：我们通过11月份已知的信息来预测当年年终该股票是否会发生高转送。
### 考虑变量(信息)
* 每股资本公积金与留存收益之和 （取不到专业版数据，用普通版季报数据替代）
* EPS(TTM)
* 总股本
* 20个交易日平均股价
* 20个交易日平均换手率
* PE
* PE1
* PB
* 每股盈余公积金
* 是否新股(上市250个交易日以内)
* 上市时间(年)
* 年中是否有分红
* 去年是否有高转送

```python
def get_point_date(start,end,baseDate='-11-01'):
    """
    根据起始和截止年度，生成各年度基准日列表
    """
    point_date = []
    for year in range(start,end):
        point_date.append(str(year)+baseDate)
    return point_date
def get_point_universe(point_date):
    """
    根据基准日列表，生成每年该基准日正在上市的股票，形成基准日和股票字典
    """
    available_universe_dict = {}
    for date in point_date:
        available_universe_dict[date] = get_universe(date)
    return available_universe_dict
```

```python
def get_factor(trade_date,universe):
    """
    根据日期和股票列表，获取因子
    """
    trade_date = cal.adjustDate(trade_date).strftime('%Y%m%d') #标准化日期
    begin_date = cal.advanceDate(trade_date,'-20B').strftime('%Y%m%d') #以当前日期以前20个交易日作为起始日期
    lastYearDate = cal.advanceDate(trade_date,'-120B').strftime('%Y%m%d') #以当前日期以前20个交易日作为起始日期
    
    ##得到EPS
    fundament = DataAPI.MktStockFactorsOneDayGet(tradeDate=trade_date,secID=universe,ticker=u"",
                                    field=u"secID,tradeDate,EPS",pandas="1")
    
    ##得到（每股资本公积金与每股留存收益之和与）、每股未分配利润、每股净资产、每股盈余公积
    filed="secID,publishDate,reserPS,rePS,nAssetPS,sReserPS,tRePS"
    #columns=['secID','公告日期','每股公积金','每股未分配利润','每股净资产','每股盈余公积','每股留存收益']
    fdmt=DataAPI.FdmtIndiPSPitGet(ticker=u"",secID=universe,endDate=trade_date,beginDate=lastYearDate,beginYear=u"",endYear=u"",reportType=u"CQ3",publishDateEnd=u"",publishDateBegin=u"",field=filed,pandas="1")
    fdmt['Total_PS']=fdmt['reserPS']+fdmt['tRePS']
    
    ##得到总股本、流通市值、PE、PE1、PB 
    filed="secID,closePrice,marketValue,negMarketValue,PE,PE1,PB"
    quotation = DataAPI.MktEqudGet(tradeDate=trade_date,secID=universe,ticker=u"",
                                   beginDate=u"",endDate=u"",isOpen="",field=filed,pandas="1")
    quotation['Total_Shares'] = quotation['marketValue'] / quotation['closePrice'] / 1e+08 ##总股本(亿股)
    
    ##计算平均股价（20日）和平均换手率（20）  
    quotation_window = DataAPI.MktEqudGet(tradeDate=u"",secID=universe,ticker=u"",
                                   beginDate=begin_date,endDate=trade_date,isOpen="",field=u"secID,closePrice,turnoverRate",pandas="1")
    quotation_window = quotation_window.groupby(['secID']).mean().reset_index().rename(columns = {'closePrice':'meanPrice','turnoverRate':'meanRate'})
    
    ##计算上市时间
    eq_list = DataAPI.EquGet(equTypeCD=u"A",secID=universe,ticker=u"",listStatusCD=u"",field=u"secID,listDate",pandas="1")
    eq_list['listYear'] = eq_list['listDate'].map(lambda x: cal.bizDatesNumber(x,trade_date)/250.0) 
    
    ##设置是否为次新（一年以内为次新股）
    eq_list['ifNew'] = eq_list['listYear'].map(lambda x:1 if x<=1 else 0)
    
    #整合数据
    summary_df = pd.merge(fundament[['secID','tradeDate','EPS']],fdmt[['secID','Total_PS','reserPS','rePS','sReserPS']],on=['secID'],how='left')
    summary_df = pd.merge(summary_df,quotation[['secID','Total_Shares','PE','PB']],on=['secID'],how='left')
    summary_df = pd.merge(summary_df,quotation_window[['secID','meanPrice','meanRate']],on=['secID'],how='left')
    summary_df = pd.merge(summary_df,eq_list[['secID','listYear','ifNew']],on=['secID'],how='inner')
    ##年中是否有过高转送
    #mid_transfer_list = high_amount[(high_amount['secID'].isin(universe)) & (high_amount['endDate']== (trade_date[0:4]+'-06-30'))]['secID'].tolist()  ##年中高转送算法不对，从数据看年中的截止日期不一定是6月30日，因此改为小于十一月份
    mid_transfer_list = high_amount[(high_amount['secID'].isin(universe)) & (high_amount['endDate']<(trade_date[0:4]+'-11-01'))&(high_amount['endDate']>(trade_date[0:4]+'-01-01'))]['secID'].tolist()
    summary_df['MidHasTransfer'] = summary_df['secID'].map(lambda x: 1 if x in mid_transfer_list else 0)
    ##去年是否有过高转送
    last_transfer_list = high_amount[(high_amount['secID'].isin(universe)) & (high_amount['endDate']== str(int(trade_date[0:4])-1)+'-12-31')]['secID'].tolist()
    summary_df['LastHasTransfer'] = summary_df['secID'].map(lambda x: 1 if x in last_transfer_list else 0)
    
    return summary_df
```

```python
def get_total_summary(universe_dict):
    """
    获取所有样本点汇总
    """
    total_summary_df = pd.DataFrame()
    for k,v in universe_dict.items():
        summary_df = get_factor(k,v)
        result = high_amount[(high_amount['publishDate']>k) & (high_amount['publishDate']<str(int(k[0:4])+1)+'-05-01')]
        summary_df['isTransfer'] = summary_df['secID'].map(lambda x:1 if x in result['secID'].tolist() else 0)
        total_summary_df = pd.concat([total_summary_df,summary_df],axis=0)
    return total_summary_df
```

## 特征处理
---
目标：对特征进行预处理，去除异常数据，并选择最优特征和模型。

```python
out_Sample_Universe = get_universe('2016-11-01')
out_Sample_df = get_factor('2016-11-01',out_Sample_Universe)
out_Sample_df = out_Sample_df.dropna()
```

```python
from sklearn.pipeline import Pipeline
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.preprocessing  import MinMaxScaler
from sklearn.preprocessing import  StandardScaler
from sklearn.feature_selection import VarianceThreshold
from sklearn.feature_selection import SelectKBest
from scipy.stats import pearsonr
from numpy import array
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression
from sklearn.feature_selection import SelectFromModel
from sklearn.ensemble import GradientBoostingClassifier

def prep(prepName='MinMaxScaler'):
    if prepName == 'MinMaxScaler':
        return MinMaxScaler()
    elif prepName == 'StandardScaler':
        return StandardScaler()
    else:
        return None
    
def fs(fsName='skb',threshold=5,k=5):
    fs = {'vt':VarianceThreshold(threshold=threshold), 'skb':SelectKBest(lambda X, Y: array(map(lambda x:pearsonr(x, Y)[0], X.T)).T, k=k),'rfe':RFE(estimator=LogisticRegression(), n_features_to_select=k),'sfm':SelectFromModel(GradientBoostingClassifier()),'pca':PCA(n_components=k)}
    return fs[fsName]
```

```python
#步骤一：标准化
step1=('prep',prep())
#步骤二：特征选择
step2=('fs',fs())
#步骤三：算法模拟
step3=('lr',LogisticRegression())

pipeline=Pipeline(steps=[step1,step2,step3])
```

```python
from sklearn.grid_search import GridSearchCV
parameters = {
    'prep': (MinMaxScaler(),StandardScaler()),
    'fs': ( SelectKBest(lambda X, Y: array(map(lambda x:pearsonr(x, Y)[0], X.T)).T, k=6),RFE(estimator=LogisticRegression(), n_features_to_select=6),SelectFromModel(GradientBoostingClassifier()),PCA(n_components=6)),
    'lr__C':[0.05,0.1, 0.2, 0.4, 0.8]
}

grid_search = GridSearchCV(pipeline, param_grid=parameters)
```

```python
from time import time
# 通过GridSearchCV来寻求最佳参数空间
print("Performing grid search...")
print("pipeline:", [name for name, _ in pipeline.steps])
print("parameters:")
print(parameters)
t0 = time()

# 这里只需调用一次fit函数就可以了
grid_search.fit(x, y)
print("done in %0.3fs" % (time() - t0))
print()

# 输出best score
print("Best score: %0.3f" % grid_search.best_score_)
print("Best parameters set:")
# 输出最佳的分类器到底使用了怎样的参数
best_parameters = grid_search.best_estimator_.get_params()
for param_name in sorted(parameters.keys()):
    print("\t%s: %r" % (param_name, best_parameters[param_name]))
```
