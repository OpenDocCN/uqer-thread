```python
import numpy as np
import pandas as pd


import cPickle as pickle
import datetime

#通联库
from CAL.PyCAL import *
cal = Calendar('China.SSE')

import numpy as np
import pandas as pd

import scipy
import math


#Ours
from lib.data_loader import load_market_data
# from lib.Alpha101Factor import alpha101_fun
from lib.Quality101Factor import  quality101_fun
def Winsorize(x):
    inplace =True
    limits  =0.01
    return  x.apply(scipy.stats.mstats.winsorize)

def zscore(x):
    mu = np.nanmean(x)
    sigma =np.nanstd(x)
    x = (x - mu) / sigma;  
    return x

def quantile_calc(series, n):
    edges = pd.Series([float(i) / n for i in range(n + 1)])
    f = lambda x: (edges >= x).argmax()
    return series.rank(pct=1).apply(f)

# 获取某日股票行业对应关系
def get_ticker_sector(securities,current_date):
    ticker_sector_df = DataAPI.EquIndustryGet(secID=securities, industryVersionCD='010303', intoDate=current_date, field='secID,secShortName,industry,industryID1,industryName1')
    ticker_sector_df['industryName1'] = [x.decode('utf8') for x in ticker_sector_df['industryName1']]
    sector_mappings = ticker_sector_df[['secID','industryName1']].copy()
    sector_mappings.set_index('secID',inplace=True)
    sector_mappings = sector_mappings['industryName1'].to_dict()
    return sector_mappings
```

# 无效因子定义

```python
market_data = load_market_data(securities,start,end,dependencies=['Open'])
non_predictive = market_data.minor_xs('Open')
non_predictive = non_predictive.pct_change(3)

non_predictive = non_predictive.stack()
non_predictive.index = non_predictive.index.set_names(['date', 'asset'])
non_predictive.name = "non_predictive"
```

# 完美因子定义

```python
lookahead_bias_days = 5

predictive5 = market_data.minor_xs('Open')
predictive5 = predictive5.pct_change(lookahead_bias_days)
# introduce look-ahead bias and make the factor predictive
predictive5 = predictive5.shift(-lookahead_bias_days)

predictive5 = predictive5.stack()
predictive5.index = predictive5.index.set_names(['date', 'asset'])
predictive5.name = "predictive5"
```

```python
lookahead_bias_days = 10

predictive10 = market_data.minor_xs('Open')
predictive10 = predictive10.pct_change(lookahead_bias_days)
# introduce look-ahead bias and make the factor predictive
predictive10 = predictive10.shift(-lookahead_bias_days)

predictive10 = predictive10.stack()
predictive10.index = predictive10.index.set_names(['date', 'asset'])
predictive10.name = "predictive10"
```

```python
#合并因子数据
factor_data = pd.concat([non_predictive,predictive5,predictive10],axis =1,join='inner')
factor_data
```

# 设置做多与做空因子

```python
Longstocks= [u'ILLIQUIDITY',]
Shortstocks = [ u'TVMA20', u'TVMA6']
```

```python
alphas = Longstocks + Shortstocks
for alpha_name in alphas:
    # if alpha_name in alpha101s:
    #     factor_df = alpha101_fun(alpha_name,pn_data)() 
    if alpha_name in uqer101s:
        factor_pn = load_uqer101_factors(securities,start,end,[alpha_name])
        factor_df = factor_pn.minor_xs(alpha_name)
    # if alpha_name in quality101s:    
    #     factor_df  = quality101_fun(alpha_name)(securities,start,end)

    if alpha_name in Shortstocks:
        factor_df = 1. / factor_df
        
    factor_df = factor_df.replace([np.inf, -np.inf,0.], np.nan)
    factor_df = factor_df.loc[start:end].copy()
    factor_df = factor_df.iloc[:-11].copy()
    factor_df = factor_df.shift().iloc[1:]

    custom_factor = factor_df.stack()
    custom_factor.index = custom_factor.index.set_names(['date', 'asset'])
    custom_factor.name  = alpha_name
    # alpha因子按市场极值化、归一化处理
    custom_factor = custom_factor.groupby(level='date').apply(Winsorize)  # 按市场去极值化
    custom_factor = custom_factor.groupby(level='date').apply(zscore)  # 按市场去极值化

    if factor_data is None:
        factor_data = pd.DataFrame(custom_factor)
    else:
        factor_data = factor_data.join(custom_factor)
factor_data.head()
```

```python
factor_data["new"] = 0.2*factor_data['TVMA20'] +0.3* factor_data['TVMA6']  #+  factor_data['ILLIQUIDITY']
```
