```python
def fig_style(ax, legend, legend_loc='upper right', xlabel=''):
    """自定义图形属性及格式：
    ax: 图形对象
    legend: 标注
    legend_loc: 标注位置
    xlable: x 坐标名称
    """
    ax.legend(legend, prop=small_font, loc=legend_loc)
    ax.title.set_font_properties(font)
    if xlabel:
        ax.set_xlabel(xlabel, fontproperties=small_font)
    ax.set_axis_bgcolor('white')
    ax.grid(color='gray', alpha=0.2, axis='y')
    return ax
```

```python
def dazong_jiaoyi():
    """大宗交易
    """
    data = DataAPI.MktBlockdGet(secID=set_universe('A'), beginDate='20160101', endDate='20160301', 
                                assetClass=u"E", field=['tradeDate', 'tradeVal'])
    data = data.groupby('tradeDate').sum()
    data.tradeVal = data.tradeVal / (10 ** 4)
    ax = data.plot(y=['tradeVal'], kind='bar', figsize=(14, 7), title=u'每日大宗交易量: (单位: 亿元)')
    fig_style(ax, [u'大宗交易量'], 'best', xlabel=u'交易日')
    
    return data

data = dazong_jiaoyi()
```
