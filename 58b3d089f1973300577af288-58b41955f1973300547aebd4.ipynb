# 自适应RNN

前几天在看GARCH时Google一下东西发现了一篇总结16年RNN的[博客](http://distill.pub/2016/augmented-rnns/)这里面提到了一个自适应计算时间的RNN，当时看的比较粗心以为是自适应RNN

DL提倡加深层数获得更好的学习效果，但是RNN模型使用目前主流还是使用tanh，这导致RNN堆砌到5层以上梯度效果变得很差，在这篇博客里面提高Alex Graves的一篇论文里面设计了一种自动调整计算步长的RNN设计，通过对序列的重复复合运算加强RNN模型效果。

总的来说就是ACT(Adaptive Computation Time for Recurrent Neural Networks)对RNN模型进行包裹，构造一个复杂的RNN模型。

对RNN和RNN变种，一个cell的接口包括 两个输入两个输出, ACT通过嵌套RNN实现自适应计算。

换而言之也就是在高度无法继续叠加的时候增加宽度，加大RNN模型的复杂性增强非线性效果。
