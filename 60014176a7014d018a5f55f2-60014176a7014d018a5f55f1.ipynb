```python
#收益率分类 每个月末交易日，前50%为1 ，后50%为0
rets=pd.DataFrame()
for d in data['calendarDate'].values[:]:
    
    
    
    lables = data0[data0['tradeDate']==d]['return']>=data0[data0['tradeDate']==d]['return'].mean()  #为分类处理数据，判断收盘价是否大于今日收盘价
    lables = lables.replace({True:'1',False:'0'})
    
    
    rets=pd.concat([rets,lables])
rets.columns=['rets']
data0=pd.concat([data0,rets],axis=1)    
df=data0
df.head()
```

```python
def get_train_val_test_data(date):
    '''
    数据集划分为训练集与测试集
    '''
    #计算前12个月，采用365天
    d_12=(datetime.datetime.strptime(date, "%Y-%m") - datetime.timedelta(days=365)).strftime('%Y-%m')
    train_df=df[(df['y_m']>=d_12) & (df['y_m']<date)]
    #train_val_df = train_val_df.sample(frac=1).reset_index(drop=True)   
    #拆分训练集、验证集                                 
    test_df =df[df['y_m']==date]   
    return train_df,test_df

def format_feature_label(origin_df, is_filter=True):
    '''
    将原始数据集划分为特征（因子）与数据标签
    '''
    df_test= origin_df                
    feature = origin_df[factor_names]
    label = origin_df['rets']

    return feature, label,df_test

```

```python
LR_model_list = []
accuracy=[]
classification=[]
final=pd.DataFrame()

for date in df['y_m'].drop_duplicates()[12:]:
    

    print('training model for %s' % date)
    train_df,test_df = get_train_val_test_data(date)
    train_feature, train_label,df_train = format_feature_label(train_df)       
    test_feature, test_label,df_test = format_feature_label(test_df)
    #随机逻辑回归筛选特征
    rlr=RandomizedLogisticRegression(selection_threshold = 0.2)
    rlr.fit(train_feature,train_label)         
    train_feature1=train_feature[train_feature.columns[rlr.get_support()]] 
    if train_feature1.empty:
        factor_max=np.where(rlr.scores_==np.max(rlr.scores_))[0]
        train_feature1=train_feature[factor_max]
    #建立模型                                              
    lr=LogisticRegression()           
    #格点搜索   
    params={'C':[0.1,1,10,100]}
    clf=GridSearchCV(lr,param_grid=params,n_jobs=1)
    clf.fit(train_feature1,train_label)
    clf_best=clf.best_estimator_
    clf_best.fit(train_feature1,train_label)

    #验证集        
    test_feature1=test_feature[test_feature.columns[rlr.get_support()]] 
    if test_feature1.empty:
        test_feature1=test_feature[factor_max]
    predict_prob=clf_best.predict_proba(test_feature1)[:,1]   
    print('accuracy : %s '%(clf_best.score(test_feature1,test_label)))
    accuracy.append(clf_best.score(test_feature1,test_label))
    LR_model_list.append(clf_best)
    df_test.loc[:,'factor'] = predict_prob
    df_test = df_test.loc[:, ['stock', 'tradeDate', 'rets', 'factor']] 
    final=pd.concat([final,df_test])
```
