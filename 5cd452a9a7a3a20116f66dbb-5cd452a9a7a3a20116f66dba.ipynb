# 北上资金净流入情况

```python
import matplotlib.pyplot as plt
import numpy as np
from datetime import datetime 

beginDate='20180101'
endDate='20190508'


#沪股通资金流入
data = DataAPI.equSHHKstatsTDGet(tradeDate=u"",exchangeCD=u"SHSC",beginDate=beginDate,endDate=endDate,field=u"tradeDate,buyValue,sellValue",pandas="1")


#深股通资金流入
data2=DataAPI.equSZHKstatsTDGet(tradeDate=u"",exchangeCD=u"SZSC",beginDate=beginDate,endDate=endDate,field=u"tradeDate,buyValue,sellValue",pandas="1")

#沪、深资金流入合并
data=data.merge(data2,on=['tradeDate'])
data['jlr']=(data['buyValue_x']-data['sellValue_x']+data['buyValue_y']-data['sellValue_y'])/100000000


x=data['tradeDate']
xdate = [datetime.strptime(d, '%Y-%m-%d').date() for d in x]

y=data['jlr']


fig,ax1=plt.subplots(nrows=1,ncols=1,figsize=(15,6))

ax1.bar(range(len(xdate)),y,color='r',linewidth=1,tick_label = x)
#避免x轴标签过于密集的显示
for label in ax1.get_xticklabels():
    label.set_visible(False)
for label in ax1.get_xticklabels()[::20]:
    label.set_visible(True)
plt.gcf().autofmt_xdate()  # 自动旋转日期标记
plt.show()





```
