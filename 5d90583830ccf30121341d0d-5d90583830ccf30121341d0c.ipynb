# macd计算

```python
'''
TA-LIB】之MACD
https://blog.csdn.net/qq_37193537/article/details/81103051
2018年07月18日 18:38:37
'''      
import  talib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
#isOpen="1" 剔除停盘数据
df=DataAPI.MktEqudAdjGet(ticker=u"600668",tradeDate=u"",beginDate=u"2019-02-28",endDate=u"2019-09-27",isOpen="1",field=u"",pandas="1")
#print df
def myMACD(price, fastperiod=12, slowperiod=26, signalperiod=9):
    ewma12 = pd.ewma(price,span=fastperiod)
    ewma60 = pd.ewma(price,span=slowperiod)
    dif = ewma12-ewma60
    dea = pd.ewma(dif,span=signalperiod)
    bar = (dif-dea) #有些地方的bar = (dif-dea)*2，但是talib中MACD的计算是bar = (dif-dea)*1
    return dif,dea,bar
df1=DataAPI.MktStockFactorsDateRangeGet(secID=u"",ticker=u"600668",beginDate=u"2019-02-28",endDate=u"2019-09-27",field=u"DIFF,DEA,MACD",pandas="1")
df1['MACD1']=df1.apply(lambda x:x.DIFF-x.DEA, axis = 1)
print df1.head()
#第一种
macd, signal, hist = talib.MACD(df['closePrice'].values, fastperiod=12, slowperiod=26, signalperiod=9)
mydif,mydea,mybar = myMACD(df['closePrice'].values, fastperiod=12, slowperiod=26, signalperiod=9)

#画图
fig = plt.figure(figsize=[18,5])
#plt.plot(df.index,macd,label='macd dif')
#plt.plot(df.index,signal,label='signal dea')
plt.plot(df1.index,df1.MACD1,label='df1-MACD1')
plt.plot(df.index,hist,label='hist bar')
#plt.plot(df.index,mydif,label='my dif')
#plt.plot(df.index,mydea,label='my dea')
plt.plot(df.index,mybar,label='my bar')
#plt.legend(loc='best')
plt.legend()
plt.show()
'''
1.可以看出Talib中的MACD对前33个初始值是未定义的，使用时需注意。

2.可以自己编程，将初始值赋值为计算的第一天价格，得到曲线，由于初始值赋值不同在一段时间后，自己定义的macd函数和Talib提供的MACD函数值一致，说明两种计算方法一致。
'''
```
