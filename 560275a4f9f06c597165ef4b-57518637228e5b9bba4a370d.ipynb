# 【 大师系列 】价值投资策略 —— 查尔斯·布兰德 

#  *价值投资策略*

* `查尔斯·布兰德斯（1942 — ）`

>   查尔斯·布兰德斯师从本杰明·格雷厄姆1974 年创立布兰德斯投资合伙公司，其管理资金规模由初始的1.3 亿美元成长至500 亿美元的规模。布兰德斯管理的全球股票基金在截止2007 年的过去20内平均回报率达到19.21%，其执掌的美国股票基金过去10 年平均回报率达21.1%。而替AGF 基金管理的的AGF 国际基金1992 年至2002 年10 年间平均收益亦达>16.62%，在所有国际型基金中排名第一。尼尔森全球基金管理人评选也持续将布兰德斯列为全球前十的国际股权基金管理人

>  查尔斯·布兰德斯的价值投资法从三个角度对股票提出要求
>1. 其一、公司要具备强力的资产负债表
>2. 其二、要具备合理的估值
>3. 其三、公司经营阶层具备诚信且持股充足。


> > `选股条件` ：


>1. 条件_1  股票的市盈率不高于市场平均值1.5 倍
>2. 条件_2  股票的股价/近四季现金流量（市现率）不高于市场平均值的1.5 倍
>3. 条件_3  股票的市净率不高于市场平均值的1.5 倍
>4. 条件_4  股票的市净率小于2.0 
>5. 条件_5  股票最近一季负债净值比小于80%
