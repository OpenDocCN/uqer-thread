## 长短期因子的区分
如何区分因子属于长期因子还是短期因子呢？
直接办法是**人工看**，但工作量稍有点大，所以期望用程序实现此功能。
长期因子：随价格缓慢变动，所以变化更慢，平缓，理论上应该更**低频率（长波）**
短期因子：随价格迅速变动，所以变化更快，陡峭，理论上应该更**高频率（短波）**
如何区分长短波？简单点哈，时间有限
1，**方差**，如果连续信号方差应该较小（猜想，实际还和波动幅度有关）
2，**均值交叉点**，曲线按照均值平移，均值为0轴，交点多少可以反映出频率大概情况。实际是因子做diff然后计算反转个数替代。
3，**相关系数**（未处理，有兴趣自研究）

## 方差
1，因子先做差分diff
2，差分后因子在一定时间内的 “相对均值标准差=标准差/均值” 作为频率的衡量
3，如果频率高于价格closeprice对应的频率，就认为比基准更高频，可以使用
4，如果低于closeprice频率，认为因子属于长期因子，不适用短期预测

```python
from scipy.stats import mstats
import pandas as pd
#股价的相对均值标准差
stk='000001.XSHE'
begin_date='20160301'
end_date='20160330'
price_data=DataAPI.MktEqudAdjGet(secID=stk, beginDate=begin_date, endDate=end_date,field='secID,tradeDate,closePrice')
#差分
price_data['closePrice']=price_data['closePrice'].diff()
#去除极值点，对方差干扰较大
price_data['closePrice']=mstats.winsorize(price_data['closePrice'].values, limits=0.05)
#标准差/均值，认为越大越高频
prine_mean_std=price_data['closePrice'].std()/price_data['closePrice'].mean()
print prine_mean_std

#因子相对均值标准差
factor_data=DataAPI.MktStockFactorsDateRangeGet(secID=stk,ticker=u"",beginDate=begin_date,endDate=end_date,field=factor_list,pandas="1")
mean_std={}
for factor in factor_data.columns:
    try:
        #差分
        factor_data[factor]=factor_data[factor].diff()
        #去极值
        factor_data[factor]=mstats.winsorize(factor_data[factor].values, limits=0.05)
        mean_std[factor]=factor_data[factor].std()/factor_data[factor].mean()
    except Exception,e:
        print e
mean_std_s=pd.Series(mean_std)
mean_std_s.sort()
print len(mean_std_s[mean_std_s>prine_mean_std*0.8])
result1=set(mean_std_s[mean_std_s>prine_mean_std].index)
print result1
```

## 交叉点法
1，对因子做差分diff（后面的因子都是指差分后的）
2，差分后因子相邻的两天，如果有正负变化，认为有一次交叉
3，统计周期内closeprce的正负变化次数为基准，差分后因子次数，如果因子高于closeprice基准则认为高频，可以采用。否则不用

```python
from scipy.stats import mstats
import pandas as pd
#股价的均值交叉次数
stk='000001.XSHE'
begin_date='20160301'
end_date='20160330'
price_data=DataAPI.MktEqudAdjGet(secID=stk, beginDate=begin_date, endDate=end_date,field='secID,tradeDate,closePrice')
price_data['closePrice']=price_data['closePrice'].diff().apply(lambda x:1 if x > 0 else -1)
price_data['closePrice1']=price_data['closePrice'].shift(1)
print ((price_data['closePrice1']*price_data['closePrice']).value_counts()[-1])

#因子相对均值标准差
factor_data=DataAPI.MktStockFactorsDateRangeGet(secID=stk,ticker=u"",beginDate=begin_date,endDate=end_date,field=factor_list,pandas="1")
mean_std={}
for factor in factor_data.columns:
    try:
        factor_data[factor]=factor_data[factor].diff().apply(lambda x:1 if x > 0 else -1)
        factor_data[factor+'1']=factor_data[factor].shift(1)
        mean_std[factor]=(factor_data[factor+'1']*factor_data[factor]).value_counts()[-1]
    except Exception,e:
        #print e
        pass
mean_std_s=pd.Series(mean_std)
mean_std_s.sort()
print len(mean_std_s[mean_std_s>prine_mean_std*0.8])
result2=set( mean_std_s[mean_std_s>prine_mean_std].index)
print result2
```

## 做成一个工具类
给出希望的周期，输出适合此周期的因子列表（其实是比此周期高频因子），采用方差法
注意：由于方法返回的是频率高于cycle对应均线的频率，使用时可能cycle=5得到的结果中去掉cycle=5*3=15的结果。
虽然严格来说**高频信号会影响低频信号**（昨日价格也会影响20日均线走势）。但是力度很弱。所以感觉比较合理的是
适合5日周期的因子=getSuitFactor(cycle=5)-getSuitFactor(cycle=5*3)-getSuitFactor(cycle=int(5/3))
个人感觉，未必对。根据使用情景决定吧

```python
import pandas as pd
from scipy.stats import mstats
import datetime
import time
from CAL.PyCAL import *

def getSuitFactor(cycle=5,secID='000001.XSHE'):
    stk=secID
    today=Date.todaysDate()
    begin_date=(today - Period('1Y')).toDateTime().strftime('%Y%m%d')
    end_date=today.strftime('%Y%m%d')
    
    #已secID为基准，输出与cycle匹配的因子列表
    price_data=DataAPI.MktEqudAdjGet(secID=stk, beginDate=begin_date,endDate=end_date,\
                                     field='secID,tradeDate,closePrice')
    price_data['closePrice']=pd.rolling_mean( price_data['closePrice'],cycle)
    price_data['closePrice']=price_data['closePrice'].diff()
    price_data['closePrice']=mstats.winsorize(price_data['closePrice'].values, limits=0.05)
    prine_mean_std=price_data['closePrice'].std()/price_data['closePrice'].mean()

    #因子相对均值标准
    factor_data=DataAPI.MktStockFactorsDateRangeGet(secID=stk,ticker=u"",\
        beginDate=begin_date,endDate=end_date,field=factor_list,pandas="1")
    mean_std={}
    for factor in factor_data.columns:
        try:
            factor_data[factor]=pd.rolling_mean(factor_data[factor],cycle)
            factor_data[factor]=factor_data[factor].diff()
            factor_data[factor]=mstats.winsorize(factor_data[factor].values, limits=0.05)
            mean_std[factor]=factor_data[factor].std()/factor_data[factor].mean()
        except Exception,e:
            print e
    mean_std_s=pd.Series(mean_std)
    mean_std_s.sort()
    result1=set(mean_std_s[mean_std_s>prine_mean_std].index)
    return list(result1)
    
```

## 因子贡献度
本是另一个帖子的一函数，利用因子的分组收益差额进行优劣判断，恰和本帖较高相关性，就拉过来吧

```python
def getGoodFactors(begin_date,end_date,factor_num,factors=[],secID_list=[]):
    #用start_date的factor为X，end_date收盘/start_end收盘=涨跌率y，计算因子贡献度，返回较大的topN
        #创建因子df:fac
    fac=DataAPI.MktStockFactorsOneDayGet(tradeDate=begin_date,secID=secID_list,field=['secID']+factors,pandas="1")
    
    #创建价格df:price,计算区间收益
    pre_price=DataAPI.MktEqudAdjGet(secID=secID_list,tradeDate=begin_date,field=u"secID,closePrice",pandas="1")
    yes_price=DataAPI.MktEqudAdjGet(secID=secID_list,tradeDate=end_date,field=u"secID,closePrice",pandas="1")
    yes_price=yes_price.rename(columns={'closePrice':'yes_closePrice'})
    price=pd.merge(pre_price,yes_price)
    
    #区间收益率
    price['rate']=((price['yes_closePrice']/price['closePrice']-1)*100)
    price=price.drop(['closePrice','yes_closePrice'],axis=1)
    traindf=pd.merge(fac,price).set_index('secID').dropna()

    #对每一个因子排序，取前20%和后20%,计算因子贡献度=前20%均收益-后20%均收益（由于相对值，故不需计算大盘前20%和后20%收益差)
    facdict={}
    for factmp in factors:
        traindf=traindf.sort(columns=factmp)
        facjc=traindf['rate'][len(traindf)*8/10:].mean()-traindf['rate'][:len(traindf)*2/10].mean()
        facdict[factmp]=abs(float(facjc))

    #字典排序，选前X位因子
	fac_series=pd.Series(facdict)
	return fac_series.nlargest(factor_num).index.tolist()
```
