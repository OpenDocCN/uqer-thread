先造两个dataframe

```python
import pandas as pd
import numpy as np

df1 = pd.DataFrame({'a':pd.Series([2,4,63,12]),'b':pd.Series([8,56,32,1,4])})
df2 = pd.DataFrame({'c':pd.Series([5,2,3]),'d':pd.Series([8,1,4]),'e':pd.Series([3,1,4])})
print df1
print df2
```

# 数据重塑

- 层次化索引有两种好用的函数：stack/unstack （col转为index/index转为col），之前也提到过，直接看例子

```python
df3 = df1.stack()
print df3

df3 = df3.unstack()
print df3
```

两个需要注意的点：1）层次化索引重塑后**NAN值会消失**；2）默认**使用外层次索引**（index里靠右边的）

- 最近总是能拿到一张张标准的dataframe表，有什么函数能方便处理表格内容吗？这里有个pivot，拿去不送

```python
pd.pivot(df2['c'],df2['d'],df2['e'])
```

pivot里按**顺序index，column，value**，恰好是dataframe三元素，一张dataframe方便定义

- 话说一排数据的重复项如何方便去除？来来来，drop_duplicates来帮你

```python
df2['f']=pd.Series([3,2,2])
print df2

print df2.drop_duplicates(['f'])
```

这里按顺序**默认取第一个为参考基准**

- 嘿！还记得上次函数映射的内容吗，那个叫啥……来着？我记得有个map……

```python
print df1['b'].map({1:'a',4:'b'})
print df1['b'].map(lambda x:x*x)
```

series中元素级别的函数应用，注意：可以**加字典或函数**

- 除了等号赋值，还有其他替换原有值的方法吗？试试replace吧，挺好用

```python
print df1['a'].replace([np.NAN,2,4],[0,1,2])
print df1['a'].replace({np.NAN:0,2:1,4:2})
```

series中元素级别的函数应用，方便替换series元素

- 轴索引有更方便的改法吗？有……

```python
df1.rename(index={0:'a',1:'b',2:'c',3:'d'},columns={'a':1,'b':2})
```

这里rename里也可以加类似于lambda的函数

- 统计里的离散化分组分析用什么函数搞呢？离散化分组有等分及分位，这里可以试试cut()和qcut()

```python
pd.cut([1,2,1,3,3,4,5,5,5,6,7,8],[1,3,5,7,9])
```

上面是最简单，最直接的方法，直接写入data和bins，就是**数据和分组规则**

```python
pd.cut([1,2,1,3,3,4,5,5,5,6,7,8],4,precision=2,right=False)
```

上面讲了些什么故事呢……有4个要素：1）data数据，一个listlike；2）4等分，很好理解；3）小数点精确到2位；4）右边改为闭区间；

```python
pd.cut([1,2,1,3,3,4,5,5,5,6,7,8],4,precision=2,right=False,labels=['a','b','c','d'])
```

外加一个要素，label标签，很多此类函数工具都有这个功能，最后qcut和cut用法差不多，读者可自行把玩

# 字符串操作

这里介绍个很有用的公式

```python
val = 's df, geg  wer, er ,ssd'
print val

valx = [x.strip() for x in val.split(',')]
print valx
```

上面完成了以**逗号隔开且去除空格**的目的，这里strip指去除空格，split指按……隔开

- 这个挺有趣，还有什么其他字符串的操作吗？别急……介绍几个常用的

```python
str1 = 'aaaabbccccddefghij'
print str1.count('a')
print str1.index('b')
print str1.find('c')
print str1.upper()
```

上面讲了3个故事……**1）数数；2）找位置；3）变大……写；**简单说一下，find和index都是找位置用的，find相对实用一点，因为如果没有找到对应元素则会返回-1，如下：

```python
print str1.find('x')
print str1.index('x')
```

- 查询某个单词存在与否用哪个函数呢？这里得用两个二连方式

```python
pd.Series(['a','ab','abc','abcd','ef']).str.contains('abc')
```

series中str的查询功能，可以方便返回bool值
