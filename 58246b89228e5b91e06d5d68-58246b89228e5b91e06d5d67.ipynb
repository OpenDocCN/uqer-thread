---
* #### 设置一段回测前时间的指数行情数据作为推断回报分布函数的基准
* #### 设置窗口期，比如8天（交易日），根据上述数据计算指数历史8天回报率
* #### 计算最近 窗口期 - 1 天的回报率，根据历史窗口期天回报率分布情况，计算第二天可能的收益率分布情况
* #### 在(0,1)区间上设置风险阈值，如果计算得到第二天有 1-风险阈值 概率回报大于0，则持有HS300ETF，否则空仓
* #### 每次调仓时把获得的新信息加入历史行情数据，作为之后推断的基准

# `参数优化`
---
* #### 多长的窗口期能平滑资本市场的短期信号，使市场表现回到历史的常态？多大的风险水平能够满足我们的风控需求，把资金安心交给程序？
* #### 为了解决这些问题，下面引入一个quick bactest模块，测试5交易日到20交易日窗口期、0.1到0.9风险水平下，策略从`2012年7月`至`2016年7月`的表现。

```python
from __future__ import division
import pandas as pd
import numpy as np


def histVaR(histSeries = pd.Series(), riskLevel = 0.01):
    return -histSeries.quantile(riskLevel)

def periodVaR(histPrice = [], frequency = '3d', riskLevel = 0.01):
    ser = pd.Series(histPrice[::int(frequency[:-1])]).pct_change()[1:]
    return histVaR(ser, riskLevel= riskLevel)

def periodSignal(account, frequency = '8d', riskLevel = 0.3):
    '''考察近期收益与历史时段对比情况，如frequency为8d，riskLevel 0.3时返回：
        历史8日回报的30%分位数 - 最近7日回报率
        可以认为第二天70%可能回报率在返回值之上
    '''
    histReturn = account.get_attribute_history('closePrice', int(frequency[:-1]))[universe[0]]
    prevReturn = histReturn[-int(frequency[:-1])] / histReturn[-1] - 1
    return (-periodVaR(account.longHist, frequency = frequency, riskLevel = riskLevel) - prevReturn)
    
def initialize(account):                   # 初始化虚拟账户状态
    account.longHist = list(iniHist.closeIndex)
    pass


def handle_data(account):                  # 每个交易日的买入卖出指令
    global __frequency
    global __riskLevel

    account.longHist += list(DataAPI.MktIdxdGet(tradeDate = account.previous_date, ticker = '000300', field = 'closeIndex').values[0])   
    # histReturn = account.get_attribute_history('closePrice', 10)[universe[0]]
    # return2days = histReturn[-3] / histReturn[-1] - 1
    # VaR3days95 = periodIndexVaR(ticker = '000300', beginDate = '20020101', 
    #                             endDate = account.previous_date, freq = '3d', riskLevel= 0.05)
    # VaR3days50 = periodIndexVaR(ticker = '000300', beginDate = '20020101', 
    #                             endDate = account.previous_date, freq = '3d', riskLevel= 0.5)
    # # # 回报有95%的可能性高于：-VaR3days95 - return2days
    # # print '回报有95%的可能性高于',-VaR3days95 - return2days
    # # # 有50%的可能性高于： -VaR3days50 - return2days
    # # print '回报有50%的可能性高于',-VaR3days50 - return2days
    if len(account.universe) == 0:
        return
    if periodSignal(account, str(__frequency)+'d', __riskLevel) > 0:
        order_pct_to(universe[0], 1)
    else:
        order_to(universe[0], 0)
    return


#**************************
# quick backtest
#**************************
start = '2012-07-05'                       # 回测起始时间
end = '2016-07-20'                         # 回测结束时间
benchmark = 'HS300'                        # 策略参考标准
universe = ['159919.XSHE']  # 证券池，支持股票和基金
capital_base = 100000                      # 起始资金
freq = 'd'                                 # 策略类型，'d'表示日间策略使用日线回测，'m'表示日内策略使用分钟线回测
refresh_rate = 1                           # 调仓频率，表示执行handle_data的时间间隔，若freq = 'd'时间间隔的单位为交易日，若freq = 'm'时间间隔为分钟

iniHist = DataAPI.MktIdxdGet(beginDate = '20100101', endDate = '20120703', ticker = '000300', field = 'tradeDate,closeIndex')
iniHist.set_index('tradeDate', inplace = True)

import quartz as qz
__frequency = 10
__riskLevel = 0.2

def quickbt(fq, rl):
    global __frequency
    global __riskLevel
    __frequency = fq
    __riskLevel = rl

    sim_params = qz.sim_condition.env.SimulationParameters(start, end, benchmark, universe, capital_base)
    idxmap, data = qz.sim_condition.data_generator.get_daily_data(sim_params)
    strategy = qz.sim_condition.strategy.TradingStrategy(initialize, handle_data)
    bt, account = qz.quick_backtest(sim_params, strategy, idxmap, data, refresh_rate = refresh_rate)
    perf = qz.perf_parse(bt, account)
    return perf
        
res = {}
for fq in range(5,21):
    print fq
    res[str(fq)] = {str(rl): quickbt(fq,rl) for rl in np.arange(0.1, 1, 0.1)}        
```
