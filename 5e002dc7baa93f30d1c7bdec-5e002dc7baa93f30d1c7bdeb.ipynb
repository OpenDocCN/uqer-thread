# 获取数据

```python
import pandas as pd
pd.set_option('expand_frame_repr', False)
def get_stock_info(tradeDay):
    df = DataAPI.MktEqudGet(secID=u"",ticker=u"",tradeDate=tradeDay,beginDate=u"",endDate=u"",isOpen="",field=u"",pandas="1")
# 选取所有ST与*ST公司
    ST_list = df.loc[df['secShortName'].str.contains('ST')]
# 选区所有B股
    B_stock_list = df.loc[df['secShortName'].str.contains('B')]
# 列出ST与B股包含的行号
    remove_list = list(set(ST_list.index.append(B_stock_list.index)))
# 删除ST与B股
    for remove_stock in remove_list:
# print remove_stock
        df = df.drop([remove_stock])
    return df

tradePeriod = pd.date_range('8/7/2018','8/8/2018').strftime('%Y%m%d')
for tradeDay in tradePeriod: 
    tradeDay = "u\"" + tradeDay + "\""
    tradeDay = tradeDay.decode("utf-8")
    print get_stock_info(tradeDay)
```
