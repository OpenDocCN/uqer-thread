  # 分级基金下折分享
  
  
   看了下uqer上的分享，觉得还挺赞的，打算尝试下以后在这里开始一些探索学习。
   作了个简单的实验，看了下折样本和整体样本在一些指标中的分布情况是否有什么规律。嗯，这是一个极其简单而没什么创造性的分析工作。所以想扯一些方法论的东西来增加点分享的重量和探讨学习的点～   
   
   有很多事件驱动的策略，大部分是事件是因，对果进行分析。而在分级基金的下折事件是一个有公共知道的因的，不像拆分股权或者分红需要内幕或者不那么显而易见的因导致的。因此下折事件驱动策略算是有别于那些事件驱动策略的。他的可预测性感觉会更强，更容易些，因此那些事件驱动策略重在事件后，下折这种应该预测更重要些。但是发现想做事件预测，但样本量太少，不管放怎样的特征进去，都没有什么效果。也许可以增加下折样本量来处理，但是没有想到什么over－sample方法（其实是没有去研究。囧）。人为捏一些下折样本。所以就从特征分布来下手了（除了分布，再分析些什么比较好？），倘若挑选的特征在事件样本和整体样本中有不同的话，有两种可能：（1）该特征可以用来预测事件；（2）事件的发生会因此特征的变化   
   暂测试如下几个特征分布：
   
   *    下折指标
   
   *     上折指标
     
   *     A端日增长率 
   
   *     A端贴水率
   
   *     母鸡溢价率
   
   *   B端贴水率
   
   *   B日增长率 
 
  结论如下：<br></br>
      1.   ［果］ 下折样本中A增长率为正的较全样本多
      2.  ［因］ 下折指标有一定的区分意义，由于下折样本量比较起少，因此只能提高覆盖率，用当前指标提高准确率就显得十分有限了
      3.   ［无］B端的贴水率在下折样本中没有明显的聚集中一个范围内，和整体样本比较起来。可能对下折事件的不管是Precision还是Recall都没有太大意义
      4.    ［无］B日增长率在两个样本中分布好像的说
      5.    ［有］比较惊讶！！B平日折价多？（和常识不太符合），下折样本中B溢价多。
      6.     [有]  A溢价率，整体样本中看，大部分时候A在折价。
