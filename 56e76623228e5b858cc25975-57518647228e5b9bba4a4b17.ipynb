# 基于“互联网+量化投资 大数据指数手把手”之个股及大盘涨幅因子

本策略基于孙大神的“互联网+量化投资 大数据指数手把手”的框架和代码修改而来，在此感谢大神的无私奉献。

策略思路：
1）中国股市整个就是一个赌场，股票涨跌不是依据其自身的价值，而是是否有人炒作，所以不考虑基本面因素。
2）机器学习目的是预测股票后期涨跌和幅度，然后选择预计最大涨幅股票买入，所以这里选用股票多个不同周期的涨幅作为预测因子。
3）考虑到股票涨跌很大程度上依赖大盘环境，将大盘指数的涨跌也必须一并考虑。为减少因子数量，将个股涨幅与大盘同周期涨幅之差作为最终的预测因子。

问题探讨：
1）感觉机器学方法只适用于单边趋势行情，牛市单边做多，熊市单边做空。震荡市和熊市中，一方面单个股票走强的持续性不好，很容易买到高点。另一方面股票很可能忽上忽下，机器学习得不到正确或有价值的信息，导致预测的不准确。
2）从回测曲线看，也印证了以上观点。大盘一旦走牛，净值曲线非常陡峭。在震荡市和熊市中，基本没有超额收益，甚至表现更差，净值回测很大，为此，本策略简单的根据大盘30日均线做择时应对。
3）看到令人咋舌的结果和异常陡峭净值曲线，自己第一时间的反应是用到了未来数据，自己仔细检查过数据，没发现，还请大神指点。从震荡市和熊市中的表现看，跌的一点都不含糊，这从侧门表明没有用到未来数据？

```python
import pandas as pd
import numpy  as np
import quartz as qz
from CAL.PyCAL import *
from datetime import datetime
import statsmodels.api as sm
import statsmodels.regression.linear_model as lm
from sklearn.linear_model import ElasticNet

used_factors = ['1returnCindex', '5returnCindex', '10returnCindex', '20returnCindex', '30returnCindex']
def StockFactorsGet(universe, trading_day):
    data_all = {}
    trading_day = Date(trading_day.year, trading_day.month, trading_day.day)
    cal = Calendar('China.SSE')
    start_date = cal.advanceDate(trading_day, '-120B', BizDayConvention.Following)
    end_date = cal.advanceDate(trading_day, '-1B', BizDayConvention.Following)  
    start_date = start_date.strftime('%Y%m%d')
    end_date = end_date.strftime('%Y%m%d')  
    #去大盘指数
    sh_index=DataAPI.MktIdxdGet(indexID="000016.ZICN",beginDate = start_date, endDate = end_date, field = ['tradeDate', 'closeIndex'])
    sh_index['tradeDate'] = pd.to_datetime(sh_index['tradeDate'])
    sh_index['1fwdIndex'] = sh_index['closeIndex'].shift(1)
    sh_index['5fwdIndex'] = sh_index['closeIndex'].shift(5)
    sh_index['10fwdIndex'] = sh_index['closeIndex'].shift(10)
    sh_index['20fwdIndex'] = sh_index['closeIndex'].shift(20)
    sh_index['30fwdIndex'] = sh_index['closeIndex'].shift(30)            
    sh_index['1return_index'] = sh_index['closeIndex'] / sh_index['1fwdIndex']-1.            
    sh_index['5return_index'] = sh_index['closeIndex'] / sh_index['5fwdIndex']-1.                        
    sh_index['10return_index'] = sh_index['closeIndex'] / sh_index['10fwdIndex']-1.
    sh_index['20return_index'] = sh_index['closeIndex'] / sh_index['20fwdIndex']-1.
    sh_index['30return_index'] = sh_index['closeIndex'] / sh_index['30fwdIndex']-1.               
    
    for i,stock in enumerate(universe):
        try:
            # 去个股涨幅
            data = DataAPI.MktEqudGet(secID = stock, beginDate = start_date, endDate = end_date, field = ['tradeDate', 'closePrice'])
            data['tradeDate'] = pd.to_datetime(data['tradeDate'])
            data['1fwdPrice'] = data['closePrice'].shift(1)
            data['5fwdPrice'] = data['closePrice'].shift(5)
            data['10fwdPrice'] = data['closePrice'].shift(10)
            data['20fwdPrice'] = data['closePrice'].shift(20)
            data['30fwdPrice'] = data['closePrice'].shift(30)            
            data['1return'] = data['closePrice'] / data['1fwdPrice']-1.            
            data['5return'] = data['closePrice'] / data['5fwdPrice']-1.                        
            data['10return'] = data['closePrice'] / data['10fwdPrice']-1.
            data['20return'] = data['closePrice'] / data['20fwdPrice']-1.
            data['30return'] = data['closePrice'] / data['30fwdPrice']-1.                    
            
            # 股票与大盘指数数据合并
            data = data.merge(sh_index, how = 'inner', left_on = ['tradeDate'], right_on = ['tradeDate'])  
            data=data[60:]            
            data['1returnCindex']= data['1return'] - sh_index['1return_index']
            data['5returnCindex']= data['5return'] - sh_index['5return_index']            
            data['10returnCindex']= data['10return'] - sh_index['10return_index']
            data['20returnCindex']= data['20return'] - sh_index['20return_index']            
            data['30returnCindex']= data['30return'] - sh_index['30return_index']                   
            data=data.loc[:, ['tradeDate']+used_factors]
           
        except Exception, e:
            print e

        data_all[stock] = data

    return data_all

def StockRegDataGet(stock, trading_day, factors, shift = 2):
    trading_day = Date(trading_day.year, trading_day.month, trading_day.day)
    cal = Calendar('China.SSE')
    start_date = cal.advanceDate(trading_day, '-60B', BizDayConvention.Following)
    end_date = trading_day        
    data=factors   
    start_date = start_date.strftime('%Y%m%d')
    end_date = end_date.strftime('%Y%m%d')    
    ret = DataAPI.MktEqudGet(secID = stock, beginDate = start_date, endDate = end_date, field = ['tradeDate', 'closePrice'])
    ret['tradeDate'] = pd.to_datetime(ret['tradeDate'])
    ret['fwdPrice'] = ret['closePrice'].shift(-shift)
    ret['return'] = ret['fwdPrice'] / ret['closePrice'] - 1.
    ret = ret[:-shift]
    data = data.merge(ret, how = 'inner', left_on = ['tradeDate'], right_on = ['tradeDate'])
    data = data.loc[:, ['tradeDate','return'] + used_factors]
    return data

def GetRegressionResult(data):
    data = data.dropna()
    all_factors = used_factors
    
    for f in all_factors:
        if data[f].std() == 0:
            continue
        data[f] = (data[f] - data[f].mean()) / data[f].std()
    
    y = np.array(data['return'].tolist())
    x = []
    for f in all_factors:
        x.append(data[f].tolist())
    x = np.column_stack(tuple(x))
    x = np.array( [ np.append(v,1) for v in x] ) 
    
    en = ElasticNet(fit_intercept=True, alpha=0)
    en.fit(x, y)
    res = en.coef_[:-1]
    w = dict(zip(all_factors, res))
    return w

def preparing(universe, trading_day, factors_all):
    datas, means, vols, weights = {}, {}, {}, {}
    for i,stock in enumerate(universe):
        try:
            datas[stock] = StockRegDataGet(stock, trading_day, factors_all[stock])
            means[stock] = dict(datas[stock].mean())
            vols[stock] = dict(datas[stock].std())
            weights[stock] = GetRegressionResult(datas[stock])
            
        except Exception, e:
            pass
    return means, vols, weights
```
