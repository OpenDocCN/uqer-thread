```python
#第一步，加载所需要的包
import numpy as np
import pandas as pd
import scipy.stats as st
from CAL.PyCAL import *     # to be commented
font.set_size(22)
small_font = font.copy()
small_font.set_size(12)
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("white")
import math
universe = set_universe("ZZ500")
bdate = "20100101"
edate = "20190315"
# factors = ["PB",'PCF','PE','ILLIQUIDITY','REVS20',"SRMI",'VOL20']
factors = ['ILLIQUIDITY','VOL20','SRMI','LCAP','CTOP','REVS20','PB']
my_factors = ['VCHL','EPS_VOL']

data = DataAPI.TradeCalGet(exchangeCD = u"XSHG", beginDate=bdate, endDate=edate, field = ["calendarDate",'isWeekEnd',"isMonthEnd","isOpen"]) #[data["isOpen"] == 1]
data = data[data["isOpen"] == 1] # get the date when market is open
tdate = data["calendarDate"].apply(lambda x: x[:10])
tdate = list(tdate)
```

```python
# import the factor's data

factors_new = ['WVAD','MACD','KDJ_K','DBCD','VMACD']
factor_data = pd.DataFrame()
f = pd.DataFrame()  # for daily storage
df_factors = {}
for factor in factors_new:
    df_factors[factor] = pd.DataFrame() #for storage of every factor's daily data (a dataframe whose index is dates and columns are names of securities)
for i in tdate:
    print i
    data = DataAPI.MktStockFactorsOneDayGet(tradeDate = i, secID = universe, field = ["tradeDate", "secID"]+factors_new) #暂时没有必要dropna，因为concat之后na又会被补上
    #接下来，每天拿到数据以后，需要把每天的因子数据拆分到每个单独的因子的list上去
    for factor in factors_new:
        mean_factor = data[factor].mean()
        data[factor] = data[factor].fillna(mean_factor)     #先fillna
        data[factor] = np.power(data[factor].rank()/len(data) - 0.5,3)
        data_factor = data[["tradeDate","secID"]+[factor]].pivot(index = "tradeDate", columns = "secID", values = factor)
        df_factors[factor] = pd.concat([df_factors[factor],data_factor],axis = 0)

for factor in factors_new:
    df_factors[factor].to_csv("yutao_"+factor+"_data.csv")
```

```python
#import data of daily openPrice
#时间上，我们回测到20190315，那么相应地，我们在20190315这一天，需要用昨天的因子数据来确定交易仓位，所以，我么的因子数据最多可以用到20190314这一天；在组合因子的时候，我们需要考虑因子过去一段时间的IC，那么每一天，我们可以知道昨天的开盘价和前天的开盘价，从而可以得到前天开始到昨天的涨幅，这部分的预测相应地是利用大前天的因子数据得到的，所以我们在20190315这一天，要用到20190314的因子数据，以及20190314的权重数据，而20190314的权重数据，应该利用20190313，20190314的open数据和20190312的因子数据得到的IC值修正。为此，我们可以在每一天给出一个过去一天的收益率，也就是今天的open-昨天的open/昨天的open。这个数据是每天开盘以后才知道的行情。
daily_open = pd.DataFrame()

for i in tdate:
    print i
    data = DataAPI.MktEqudAdjGet(secID=universe,tradeDate=i,beginDate=i,endDate=i,isOpen="",field=["tradeDate","openPrice","secID"],pandas="1")     #get the open price for today
    data_open = data.pivot(index = "tradeDate", columns = "secID", values = "openPrice")    # 得到今日的单行数据
    daily_open = pd.concat([daily_open, data_open],axis=0) #拼接起来得到每日的open序列
daily_open_return = daily_open / daily_open.shift() - 1
daily_open_return[np.isinf(daily_open_return)] = np.nan
daily_open.to_csv('yutao_open_data.csv')
del daily_open
daily_open_return.to_csv("yutao_open_return_data.csv")
```

```python
daily_close = pd.DataFrame()

for i in tdate:
    print i
    data = DataAPI.MktEqudAdjGet(secID=universe,tradeDate=i,beginDate=i,endDate=i,isOpen="",field=["tradeDate","closePrice","secID"],pandas="1")     #get the open price for today
    data_close = data.pivot(index = "tradeDate", columns = "secID", values = "closePrice")    # 得到今日的单行数据
    daily_close = pd.concat([daily_close, data_close],axis=0) #拼接起来得到每日的open序列
daily_close_return = daily_close / daily_close.shift() - 1
daily_close_return[np.isinf(daily_close_return)] = np.nan
daily_close.to_csv('yutao_close_data.csv')
del daily_close
daily_close_return.to_csv("yutao_close_return_data.csv")
```

```python
daily_vol = pd.DataFrame()

for i in tdate:
    print i
    data = DataAPI.MktEqudAdjGet(secID=universe,tradeDate=i,beginDate=i,endDate=i,isOpen="",field=["tradeDate","turnoverVol","secID"],pandas="1")     #get the volume for today
    data_vol = data.pivot(index = "tradeDate", columns = "secID", values = "turnoverVol")    # 得到今日的单行数据
    daily_vol = pd.concat([daily_vol, data_vol],axis=0) #拼接起来得到每日的vol序列
daily_vol_change = daily_vol / daily_vol.shift() - 1
daily_vol_change[np.isinf(daily_vol_change)] = np.nan
daily_vol.to_csv('yutao_vol_data.csv')
del daily_vol
daily_vol_change.to_csv("yutao_vol_change_data.csv")
```

```python
daily_open_return = pd.read_csv("yutao_return_data.csv").set_index("tradeDate")    # read the return data for today
daily_open_return[np.isinf(daily_open_return)] = np.nan
# Note： 每天使用的weight设定为是供当天使用的，因此可以用到的数据是：today-1的daily_open_return,以及today-3的factor数据
#print factor_dic
######开始计算IC
IC_data=pd.DataFrame()
#现在factor的DataFrame 和 daily_open_return 的DataFrame都已经有了，那么就把两个DataFrame横向地concat起来，然后再求corr
for factor in factors+my_factors:
    factor_pd = pd.read_csv("yutao_"+factor+"_data.csv").set_index("tradeDate")     # 先读取数据
    IC = pd.concat([factor_pd.shift(3),daily_open_return.shift(1)],axis = 1)    #横向地拼接起来，每一个index对应三天前的factor和昨天的return
    size = int(len(factor_pd.columns))
    del factor_pd     #释放内存
    IC = IC.apply(lambda row: row[:size].corr(row[size:]) ,axis = 1)    #得到factor因子的IC序列啦
    IC_data[factor] = IC #存入IC_data里面
IC_data.to_csv("IC_data.csv")
print IC_data.tail()
```
