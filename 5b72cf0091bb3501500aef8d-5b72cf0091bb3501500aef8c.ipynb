# 简单的多因子流程

多因子选股模型是相当简单易懂的一种量化投资框架。每次我们都会尝试各种创建因子，使用各种加权方式，所以我尝试做成一个标准化的流程（尽管优矿好像自带了因子选股，但我还是尝试着写一下）。中间步骤如果有问题欢迎提出，我及时改正
这里我选择从股票的成长性、估值、盈利性、财务结构、现金流充裕度几个方面来选择股票

```python
import pandas as pd
import numpy as np
import statsmodels.api as st
import matplotlib.pyplot as mpt
import scipy as sc
stocks=set_universe('HS300')            #选股的股票池
beg='20100101'            #起始日期
end='20180101'            #结束日期
#这里填的因子是来自于优矿API.MktStockFactorsOneDayGet的。如果是其他因子，可能还需要自行计算
factors=u"secID,tradeDate,PE,PB,PS,LCAP,NetProfitGrowRate,NetAssetGrowRate,DEGM,OperatingRevenueGrowRate,OperatingProfitGrowRate,GrossIncomeRatio,NetProfitRatio,ROE,ROA,ETOP,TotalAssetsTRate,CurrentAssetsTRate,IntangibleAssetRatio,BLEV,DebtsAssetRatio,FixAssetRatio,NOCFToOperatingNI"
alpha=0.05   #在利用相关性选取因子时，选择的p_value的临界值
method='pearsonr' #决定使用何种相关性判断方法。pearsonr还是spearmanr
q=0.7       #分位数值。决定你将赋予因子值前百分之多少的股票1分，后百分之多少的股票-1分
type_weight=[0.25,0.2,0.3,0.15,0.1]    #5种维度的因子赋予的权重的大小
```

由于我尝试的几乎都为基本面因子，因此我只在每个月末或者季末才计算一次因子值。由于我想要计算IC值，因此我还需要获取下个月的股票收益。这里的代码虽然也收集了股票的行业分类，但是本次研究中并没有用到。

```python
#############################获取日历，月度
def get_cal(begs=beg,ends=end):
    date=DataAPI.TradeCalGet(exchangeCD=u"XSHG",beginDate=(pd.to_datetime(begs)-pd.offsets.MonthEnd()-5*pd.offsets.Day()).strftime('%Y%m%d'),endDate=ends,field=u"calendarDate,isMonthEnd,isQuarterEnd",pandas="1")
    date=date[date.isMonthEnd==1]
    date['prevTradeMonth']=date['calendarDate'].shift(1)
    date1=date[date.isQuarterEnd==1]
    date1['prevTradeQuarter']=date1.calendarDate.shift(1)
    date=pd.merge(date,date1[['calendarDate','prevTradeQuarter']],on='calendarDate',how='outer')
    return date[date.prevTradeMonth.isnull()!=True]

p=DataAPI.MktRankDivYieldGet(beginDate=beg,endDate=end,secID=stocks,ticker=u"",field=u"secID,tradeDate,divRate",pandas="1")
#############################估值因子：PE,PB,PS,市值   type:TTM
#############################成长因子：净利润，净资产，毛利，营收，营业利润的增长率
#############################盈利因子：毛利，净利润，ROE,ROA，营业收入/市值（计算）
#############################结构因子：资产周转率，流动资产周转率，无形资产比率，非流动杠杆（账面杠杆），总杠杆，固定资产比率
#############################现金因子：经营现金净利润比率（计算），股息率（其他地方找）
date=get_cal(beg,end)
df=pd.DataFrame()
for day in date.calendarDate:
    df=df.append(DataAPI.MktStockFactorsOneDayGet(
        tradeDate=day,secID=stocks,ticker=u"",
        field=factors,pandas="1"))

df.ETOP=df.ETOP/df.NetProfitRatio
df.NOCFToOperatingNI=df.NOCFToOperatingNI*df.NetProfitRatio
############加上股息和行业分类
def get_ind(dataf=df):
    ind=DataAPI.EquIndustryGet(secID=dataf.secID.unique(),ticker=u"",industryVersionCD=u"010303",
                               industry=u"",industryID=u"",industryID1=u"",
                               industryID2=u"",industryID3=u"",intoDate=u"",
                               field=u"secID,isNew,industryName1,industryName2",pandas="1")
    ind=ind[ind.isNew==1].drop('isNew',axis=1)
    return pd.merge(dataf,ind,how='left',on='secID')

df=get_ind(df)
df=pd.merge(df,p,how='left',left_on=['secID','tradeDate'],right_on=['secID','tradeDate'])
df=pd.merge(df,date,how='left',left_on='tradeDate',right_on='calendarDate')
#拿下一月前复权的月行情
def get_ret(dataf=df,begs=beg,ends=end):
    ret=DataAPI.MktEqumAdjGet(
        secID=dataf.secID.unique(),ticker=u"",monthEndDate="",beginDate=(pd.to_datetime(begs)-pd.offsets.MonthBegin()).strftime('%Y%m%d'),endDate=ends,isOpen=u"",
        field=u"secID,endDate,return",pandas="1")
    return pd.merge(dataf,ret,how='left',left_on=['secID','prevTradeMonth'],right_on=['secID','endDate'])

df=get_ret(df)
del p
df.prevTradeQuarter.bfill(inplace=True)
df.groupby('prevTradeQuarter').mean()
df.prevTradeQuarter.bfill(inplace=True)
df_quarter=df.groupby(['prevTradeQuarter','secID']).mean().dropna()
column=df.columns[2:23]+df.columns[25:-7]
df.dropna(inplace=True)
factor_type=pd.Series([1]*4+[2]*5+[3]*5+[4]*6+[5]*2,index=column)
```

这里理论上还要对异常值进行处理，但是优矿的因子值好像已经完成了这一步因此就没有在意。
因子值计算完毕后我们简单查看一下因子值和收益的相关性，分别进行pearsonr和spearmanr计算。当然也可以计算和其他东西的相关性，比如对数市值

```python

def show_corr_pea(a=df_quarter,b=df_quarter['return']):
    d=[]
    for factor in column:
        d.append(sc.stats.pearsonr(a[factor],b))
    d=pd.DataFrame(d,index=column,columns=['corr','p_value'])
    #print('factors corr_pear:')
    #print(d.sort_values('p_value').to_html())
    return d.sort_values('p_value')

def show_corr_spea(a=df_quarter,b=df_quarter['return']):
    d=[]
    for factor in column:
        d.append(sc.stats.spearmanr(a[factor],b))
    d=pd.DataFrame(d,index=column,columns=['corr','p_value'])
    #print('factors corr_spear:')
    #print(d.sort_values('p_value').to_html())
    return d.sort_values('p_value')

pd.merge(show_corr_pea(),show_corr_spea(),right_index=True,left_index=True,suffixes=['_pear','_spear'],how='inner')
```

当然这里计算两个系数是为了筛选因子。我仅选取p_value小于0.05的因子作为有效因子。不过这里我是选取整个周期上的数值进行计算，我觉得选取一段一段时间的因子值计算可能会更好。

```python
def corr_pea_select(a=df_quarter,b=df_quarter['return'],alpha=0.05):
    d=[]
    for factor in column:
        d.append(sc.stats.pearsonr(a[factor],b))
    d=pd.DataFrame(d,index=column,columns=['corr','p_value'])
    return d[d.p_value<alpha]

def corr_spea_select(a=df_quarter,b=df_quarter['return'],alpha=0.05):
    d=[]
    for factor in column:
        d.append(sc.stats.spearmanr(a[factor],b))
    d=pd.DataFrame(d,index=column,columns=['corr','p_value'])
    return d[d.p_value<alpha]

if (method=='spearmanr'):
    column_select=corr_spea_select(alpha=alpha)
else:
    column_select=corr_pea_select(alpha=alpha)
```

rank评分模型（pearsonr）。首先我们已经对这些因子进行排序，给予每个因子前30%的股票1分，后30%-1分。另外这里还需要考虑因子的方向问题进行调整，你可以选择通过直觉调整，也可以用过计算的相关系数调整
这里简单地按照5分位数计算收益率。并且将各组收益率和总体的收益率均值进行对冲，画出图像。
由于想要把各组的数值统一在1开始画，所以在每一组前面加了一个1，index日期是随便写的，不用在意

```python
####################rank模型评分.这里要注意因子的方向问题
#
#############################估值因子：PE,PB,PS,市值   个人认为，都为正。市值为负
#############################成长因子：净利润，净资产，毛利，营收，营业利润的增长率     均为正
#############################盈利因子：毛利，净利润，ROE,ROA，营业收入/市值（计算）    均为正
#############################结构因子：资产周转率，流动资产周转率，无形资产比率，非流动杠杆，总杠杆，固定资产比率   正，杠杆小，固定资产小
#############################现金因子：经营现金净利润比率（计算），股息率（其他地方找） 均为正

df1=df.copy().set_index(['tradeDate','secID'])
df1[column_select.index.values]=df1[column_select.index.values]*np.sign(column_select['corr'])
column_select=column_select.index.values
#计算一下权重。
t=(factor_type[column_select].value_counts().sort_index()+pd.Series([0]*5,index=range(1,6))).fillna(0)
weight=type_weight/t
weight[np.isinf(weight)]=0
f=factor_type[column_select].replace(weight.to_dict())
top=df1.groupby(as_index=True,level=0)[column_select].quantile(q)
last=df1.groupby(as_index=True,level=0)[column_select].quantile(1-q)
df1=pd.merge(df1.reset_index(),top,how='left',left_on='tradeDate',right_index=True,suffixes=('','_top'))
df1=pd.merge(df1.reset_index(),last,how='left',left_on='tradeDate',right_index=True,suffixes=('','_last'))
df1.drop('index',axis=1,inplace=True)
a=DataAPI.MktIdxmGet(beginDate=beg,endDate=end,indexID=u"",ticker=u"000300",field=u"endDate,chgPct",pandas="1").set_index('endDate')
df1=pd.merge(df1,a,how='left',left_on='tradeDate',right_index=True)
df1.set_index(['tradeDate','secID'],inplace=True)

def fun(x):
    y=x.sort_values(ascending=0)
    y[:int(len(y)*0.2)]=1
    y[int(len(y)*0.8):]=-1
    y[int(len(y)*0.2):int(len(y)*0.8)]=0
    return y

df2=pd.DataFrame()
for factor in column_select:
    df2[factor]=df1[factor].groupby(as_index=True,level=0).apply(fun)
    
df2['score']=(df2[column_select]*f).sum(axis=1)
df2.reset_index(level=0,drop=True,inplace=True)
df2=df2.groupby(as_index=True,level=0).apply(lambda x:x.sort_values('score',ascending=0)).reset_index(level=0)
df2['return']=df.set_index(['tradeDate','secID'])['return']
df2['benchmark']=df1['chgPct']

#简单计算收益率
stra=pd.DataFrame()
for i in range(1,6):
    stra['No.%s' %i]=df2.groupby(as_index=True,level=0).apply(lambda x:x[int(len(x)/5*(i-1)):int(len(x)/5*i)]['return'].mean()-x['return'].mean())
line=pd.DataFrame(columns=stra.columns,index=['2015-01-01'])
line.ix['2015-01-01']=[0]*5
line=line.append(stra)
#这2个是没有考虑成本的收益率
(line+1).cumprod().plot(linewidth=3)
mpt.title('without cost and cumpord')
line.cumsum().plot(linewidth=3)
mpt.title('without cost and cumsum')
#这2个是考虑了成本的收益率图像。成本取千分之1.3（千分之1印花税+万分之3手续费）
(line+1-0.0013).cumprod().plot(linewidth=3)
mpt.title('with cost and cumpord')
(line-0.0013).cumsum().plot(linewidth=3)
mpt.title('with cost and cumsum')
```

3.这里只考虑了相关性的方向和p_value阈值，并没有考虑相关性的大小。权重方面我是根据个人偏好进行设置的，这个参数怎么确定，有待考究
