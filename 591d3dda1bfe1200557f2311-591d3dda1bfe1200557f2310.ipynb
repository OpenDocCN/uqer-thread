# ReportAnalyze

```python
import numpy as np
import pandas as pd
from datetime import datetime
from datetime import timedelta
import re
```

```python
pool = pd.read_excel(u'质押物池.xls',converters = {0:str})
pool.info()
```

```python
now = datetime.now()
now = now.strftime('%Y%m%d')

beginDate = (datetime.now() - timedelta(days = 1)).strftime('%Y%m%d')
```

```python
pool_list = pool[u'标的股票代码'].values
```

```python
df = DataAPI.ReportContentGet(ticker=pool_list,beginDate=beginDate,endDate=u"",field=u"",pandas="1")
df = df.sort_values(by = 'insertTime',ascending = False)
df.dropna(inplace = True)
df.head()
```

```python
ticker2name = DataAPI.EquGet(equTypeCD=u"A",secID=u"",ticker=pool_list,listStatusCD=u"",field=u"",pandas="1")
tickername = ticker2name[['ticker','secShortName']]
tickername.head()
```

```python
df = pd.merge(df,tickername,on = 'ticker',how = 'left')
df.head()
```

```python
df.info()
```

```python
risk_words = [u'风险',u'立案',u'调查',u'退市',u'吸收合并',u'处罚',u'冻结',\
             u'离职',u'减持',u'转让',u'配股',u'配债',u'业绩承诺']
```

```python

df.info()
```

```python
df1 = df.copy()
def mapf(x,riskword):
    x = str(x).decode('utf-8')
    a = x.count(riskword)
    a = int(a)
    return a
for riskword in risk_words:
    df1[riskword] = df.txtContent.apply(mapf,args = (riskword,))
```

```python
df1.head(5)
```

```python
df2 = df1.copy()
def senf(x,riskword):
    x = str(x).decode('utf-8')
    
    pattern = re.compile(ur'.*?%s.*?。'%riskword)
    sentences = pattern.findall(x)
    total_s = str()
    for s in sentences:
        
        w1 = 0
        w2 = s.find(riskword)
        long = len(riskword)
        w = True
        m = str()
        while w2 != -1:
            m = m + s[w1:w2]+u'【'+s[w2:w2+long]+u'】'
            w1 = w2 + long
            w2 = s.find(riskword,w1)
        m = m + s[w1:]

        total_s = total_s + u'\n'+ m
    return total_s

for riskword in risk_words:
    df2[riskword+u'内容'] = df.txtContent.apply(senf,args = (riskword,))
```

```python
df2.head(10)
```

```python
short_df = df2.drop(['txtContent'],axis = 1)
```

```python
short_df.head()
```

```python
short_df.to_excel('short_df.xlsx',encoding='GB18030')
```
