下面再复杂一点，将指数和股票行情显示在一起

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
class Display:
    def disp_stock(self,stockID):
        stock_list =[]
        stock_list.append(stockID)
        raw_data = DataAPI.MktEqudGet(secID=stock_list, beginDate=self.begin, endDate=self.end, pandas='1')
        df = raw_data[['secID','tradeDate','secShortName','openPrice','highestPrice','lowestPrice','closePrice','turnoverVol']]
        dat = df[df['secID'] == stockID].set_index('tradeDate')['closePrice']
        self.firstPrice=dat[0]
        print self.firstPrice
        dat.plot(title="Close Price of 600104 with index")
    def disp_index(self,ID):
        stock_list =[]
        stock_list.append(ID)
        raw_data =DataAPI.MktIdxdGet(indexID=stock_list, beginDate=self.begin, endDate=self.end, pandas='1')
        df = raw_data[['indexID','tradeDate','secShortName','openIndex','highestIndex','lowestIndex','closeIndex','turnoverVol']]
        dat = df[df['indexID'] == ID].set_index('tradeDate')['closeIndex']
        ratio=dat[0]/self.firstPrice
        dat2=dat.apply(lambda x:x/ratio)
        dat2.plot(title="")
    
    def __init__(self,begin,end):
        self.begin=begin
        self.end=end
        #显示放大一点
        plt.figure(figsize=(12,9), dpi=80) 
disp=Display('20150101','20151231')
disp.disp_stock('600104.XSHG')
disp.disp_index('000001.ZICN')
```
