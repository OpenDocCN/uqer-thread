## 2. 基金池的设计
---
既然是选基模型，首先要解决的就是池子的问题，选股有股票池，选基对应的应该有基金池。基金池的定义取决自我们跟踪的基准，假设我们跟踪的是偏股基金（930950.CSI），那么它的成分是股票范围在60%以上的开放基金。我们定义严一点，池子定义为成立至少满1年的普通股票型，偏股混合型和灵活配置型，且连续三个报告期，股票仓位超过70%，规模超过1亿的A类基金。

```python
import numpy as np
import pandas as pd
from datetime import datetime, timedelta
```

```python
# 获取基金信息
df = DataAPI.FundGet(category=u"E,H",idxID=u"",idxTicker=u"",operationMode=u"O",beginDate="",endDate="",status="",field=u"",pandas="1")
```

```python
# 选择A类
df = df[(df['isQdii'] == 0) & (df['isFof'] == 0) & (pd.isnull(df['indexFund'])) & (pd.isnull(df['etfLof']))].sort_values(by='secShortName')
df['name'] = df['secShortName'].apply(lambda x: x.split('-')[0])
df = df.drop_duplicates(subset=['name'], keep='first')

df = df[~((df['name'].str.contains('持有')) | ((df['name'].str.contains('LOF')))) ]
```

由于连续三个报告期股票仓位超过70%，这是一个动态的条件，涉及到我们回测，所以这里我们先不做具体实现。另外，基金的交易的费用和持有时间长短有关系，因此我们这里一般持有3个月，调仓时间设置为1月，4月，7月，10月后的第一个交易日，一年换4次。

```python
def get_fund(date, all_fund_info):
    """
    给定日期，并按照所有的基金信息，筛选成立至少满1年的普通股票型，偏股混合型和灵活配置型的基金
    """
    pre_1y = str(int(date[:4]) - 1) + '-' + date[4:6] + '-' + date[6:8]
    current_fund = (all_fund_info[(all_fund_info['establishDate'] <= pre_1y) & (pd.isnull(all_fund_info['expireDate']) | (all_fund_info['expireDate'] > date))]).copy()
    
    return current_fund
```

另外我们需要一个函数告诉我们，给定任意一个日期，它的前三个会计日期是多少

```python
def get_report_date_list(date, lag):
    pre_date = pd.to_datetime(date) - timedelta(31*lag * 3)
    date_list = pd.date_range(pre_date, date, freq='m').strftime('%Y%m%d')
    
    date_list = [i for i in date_list if i[4:6] in ('03', '06', '09', '12')]
    
    return date_list[-lag:]
```

我们可以尝试筛选今天符合条件的基金有哪些

```python
current_fund_info = get_fund('20230213', df)
```

```python
current_fund_info = get_fund('20230213', df)

date_list = get_report_date_list('20230213', 3)

fund_assets = dict()
fund_assets = {
    'r1': DataAPI.FundAssetsGet(secID=current_fund_info['secID'].tolist(), reportDate=date_list[0], field=u"secID,netAsset,equityRatioInTa",pandas="1"),
    'r2': DataAPI.FundAssetsGet(secID=current_fund_info['secID'].tolist(), reportDate=date_list[1], field=u"secID,netAsset,equityRatioInTa",pandas="1"),
    'r3': DataAPI.FundAssetsGet(secID=current_fund_info['secID'].tolist(), reportDate=date_list[2], field=u"secID,netAsset,equityRatioInTa",pandas="1") 
}
```

```python
def get_available_fund_info(fund_info, fund_assets):
    r1 = fund_assets['r1'][fund_assets['r1']['equityRatioInTa'] >= 70]['secID']
    r2 = fund_assets['r2'][fund_assets['r2']['equityRatioInTa'] >= 70]['secID']
    r3 = fund_assets['r3'][(fund_assets['r3']['equityRatioInTa'] >= 70) & ((fund_assets['r3']['netAsset'] >= 1e8))]['secID']
    
    fund_list = pd.Index(r1).intersection(pd.Index(r2)).intersection(r3).tolist()
    
    return (fund_info[fund_info['secID'].isin(fund_list)]).copy()
```
