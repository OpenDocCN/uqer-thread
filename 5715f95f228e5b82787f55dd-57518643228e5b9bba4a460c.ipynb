```python
import pandas as pd
sids = ('159915.XSHE','159920.XSHE','510050.XSHG','510300.XSHG','510500.XSHG','510900.XSHG','510880.XSHG','511010.XSHG')
data = {}
for sid in sids :
    one = DataAPI.MktFunddAdjGet(secID=sid,beginDate='20150101',endDate='20160310',field= 'closePrice')
    data.update({sid : one['closePrice']})

p = pd.DataFrame(data)
p.corr()
    
```
