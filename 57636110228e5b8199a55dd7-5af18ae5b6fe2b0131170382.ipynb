我现在的基本思路是，随机分布抽取几个交易日，记录当时的因子和价格，然后观察20个交易日后的价格相比20天前是涨还是跌。然后观察因子的数值和涨跌是否有影响

```python
import pandas as pd
from CAL.PyCAL import *
universe = set_universe('HS300')
fac1 = DataAPI.MktStockFactorsOneDayGet(tradeDate=u"20130801", secID=universe, field=[u"secID", u"EPS"], pandas="1")
price1 = DataAPI.MktEqudAdjGet(tradeDate=u"20130801", secID=universe, field=[u"secID", u"closePrice"], pandas="1")
fac1 = pd.merge(fac1,price1)

fac2 = DataAPI.MktEqudAdjGet(tradeDate=u"20130830", secID=universe, field=[u"secID", u"closePrice"], pandas="1")
fac2['closePrice2'] = fac2['closePrice']
del fac2['closePrice']
fac1 = pd.merge(fac1, fac2)
fac1['increse'] = (fac1['closePrice2'] - fac1['closePrice']) / fac1['closePrice']
del fac1['closePrice']
del fac1['closePrice2']
fac1.set_index('secID', inplace=True)
fac1 = fac1.sort(columns='EPS', axis=0, ascending=True)  # 根据EPS排序，图表中X轴是按照EPS从小到大排列
fac1.plot(ylim=[-0.3,1]) 
```
