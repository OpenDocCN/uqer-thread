假如某只股票有1000万股:
第一天平均交易价格10元，
第二天换手率20%，平均交易价格11元，那第二天的筹码分布为，10元：1000 * （1 - 0.2）=800万，11元：1000 * 0.2 = 200万。
第三天换手率30%，平均交易价格12元，那第三天的筹码分布为，10元：1000 * （1 - 0.2）* （1 - 0.3）= 560万，11元：1000 * 0.2 * （1 - 0.3） = 140万，12元：1000 * 0.3 = 300万，以此类推
0天：p0=10 t0=1
1天：t1=0.20 p1=11  筹码分布： p0(10):(1-t1)*t0=0.8，               p1(11)： t1=0.2
2天：t2=0.30 p2=12  筹码分布： p0(10):(1-t1)*(1-t2)*t0=0.56,        p1(11):(1-t2)*t1=0.14  ,             p2(12) :t2=0.3
3天：t3=0.40 p3=13  筹码分布： p0(10):(1-t1)*(1-t2)(1-t3)*t0=0.336, p1(11):(1-t2)(1-t3)*t1=0.084,          p2(12):(1-t3)*t2=0.18,  p3(13) :t3=0.4
注：按【2】：历史换手衰减系数：它是一个常数参数，我们用来赋予今天换手率，也既是当日被移动的成本的权重。如果今天的换手率是A，衰减系数是n，那么我们计算昨日的被移动的筹码的总量是A*n,如果n取值为1，就是一般意义上理解的今天换手多少，就有多少筹码被从作日的成本分布中被搬移；如果n是2，那么我们就放大了作日被移动的筹码的总量..这样的目的在于突出“离现在越近的筹码分布其含义越明显”。每日成本算法是一个移动平均过程： 公式是：当日成本（换手率历史换手衰减系数）+上一日成本分布图（1-换手率历史换手衰减系数）

```python
import pandas as pd
def calcuJUN(dateT,highT, lowT, volT, minD=0.01):
    x =[]
    l = (highT - lowT) / minD
    for i in range(int(l)+1):
        x.append(round(lowT + i * minD, 2))
    #print "len(x)",len(x),"x:",x
    length = len(x)
    eachV = volT/length
    if 1.0*(volT-length*eachV)/volT>=minD:
        print "为何均布分布误差，大于最小区间:",minD
    res= pd.DataFrame(x,columns = ['price'])
    res["vol"]=eachV
    res["tradeDate"]=dateT
    res=res[["tradeDate","price","vol"]]
    return res
#res=calcuJUN(dateT="2022-04-11",highT=15.72, lowT=14.57, volT=7825460,minD=0.01)
#print res
```

```python
import pandas as pd
df = pd.DataFrame(columns=["tradeDate","mean","t","close"])
df.loc[0]=["2022-04-11",10,0,10.5]
df.loc[1]=["2022-04-12",11,0.2,10.50]
df.loc[2]=["2022-04-13",12,0.3,13.50]
df.loc[3]=["2022-04-14",13,0.4,12.50]
df['t'][0] = 0
df.set_index(['tradeDate'],inplace=True)
print df
df['1_t'] = 1 - df['t']
df['2_t'] = df['1_t'][::-1].values
df['3_t'] = df['2_t'].shift(periods = 1)
df['3_t'][0] = 1
#print(df[['1_t', '2_t', '3_t']])
df['4_t'] = df['3_t'].cumprod()[::-1].values
df['t'][0] = 1
df['chouma'] = df['t'] * df['4_t']
print df
close=df["close"]
def winner_core(df,close):
    close_price = close[-1]
    df['t'][0] = 0
    df['1_t'] = 1 - df['t']
    df['2_t'] = df['1_t'][::-1].values
    df['3_t'] = df['2_t'].shift(periods = 1)
    df['3_t'][0] = 1
    #print(df[['1_t', '2_t', '3_t']])
    df['4_t'] = df['3_t'].cumprod()[::-1].values
    df['t'][0] = 1
    df['chouma'] = df['t'] * df['4_t']
    return df.loc[df['mean'] < close_price] ['chouma'].sum()
def winner(df,close):
	result = []
	n = len(close)
	for i in range(n):
		res = winner_core(df,close[: i + 1])
		result.append(res)
	return pd.Series(result, index = close.index)
print winner(df,close)
```
