# alpha,beta,波动率，信息比率，夏普率的算法

```python
from sklearn import linear_model
import numpy as np
import math
diabetes_X_training = [[0.0073],[-0.0242],[0.0197],[-0.0355],[0.0044],[0.0325],[-0.0106],[-0.0323]]
diabetes_y_training = [0,0,0,0,0,0.0455,-0.0025,-0.0515]

lenght = len(diabetes_y_training)
beta = range(lenght)
alpha = range(lenght)
volatility = range(lenght)
sharp = range(lenght)
IR = range(lenght)
strategy_annualized_returns = range(lenght)
benchmark_annualized_returns = range(lenght)
difference = range(lenght)
sum_X = 1
sum_Y = 1
for i in range(lenght):
    sum_X = sum_X * (diabetes_X_training[i][0] + 1)
    sum_Y = sum_Y * (diabetes_y_training[i] + 1)
    strategy_annualized_returns[i] = (sum_Y-1)/(i+1)*250
    benchmark_annualized_returns[i] = (sum_X-1)/(i+1)*250
    #strategy_annualized_returns[i] = math.pow(math.pow(sum_Y,1.000001/(i+1)),250)-1
    #benchmark_annualized_returns[i] = math.pow(math.pow(sum_X,1.000001/(i+1)),250) -1
    difference[i] = strategy_annualized_returns[i] - benchmark_annualized_returns[i]
    regr = linear_model.LinearRegression()
    regr.fit(diabetes_X_training[:i+1],diabetes_y_training[:i+1])
    beta[i] = regr.coef_[0]
    alpha[i] = strategy_annualized_returns[i] - beta[i] * benchmark_annualized_returns[i]
    volatility[i] = np.std(diabetes_y_training[:i+1]) * math.sqrt(250)
    if volatility[i] == 0:
        sharp[i] = None
    else:
        sharp[i] = strategy_annualized_returns[i] / volatility[i]
    if i == 0:
        IR[i] = None
    else:
        IR[i] = difference[i] / np.std(difference[:i+1])
    
print 'alpha:',alpha,'\n beta:',beta,'\n volatility:',volatility,'\n sharpe ratio:',sharp,'\n IR:',IR
```
