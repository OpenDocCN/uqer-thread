```python
import pandas as pd

# 1、单只股票的因子数据，'AdminiExpenseRate','ROE','ASSI'
factor_frame = DataAPI.MktStockFactorsDateRangeGet(secID=u"",ticker=u"000001",beginDate=u"20160616",endDate=u"20170616",
                                                   field=['ticker','tradeDate','AdminiExpenseRate','ROE','ASSI'],pandas="1")
#获取一只股票历史上某一时间段的因子数据
#AdminiExpenseRate    管理费用与营业总收入之比（Administrative expense rate）。计算方法：管理费用与营业总收入之比=管理费用（TTM）/营业总收入（TTM）。属于质量类因子。数据更新时间：T日21点30分
#ROE	权益回报率（Return on equity）。计算方法：权益回报率=净利润（TTM）/股东权益。属于质量类因子。数据更新时间：T日21点30分
#ASSI   对数总资产（Natural logarithm of total assets）。计算方法：对数总资产=总资产的对数。属于基础科目与衍生类因子。数据更新时间：T日21点30分

print factor_frame.head().to_html()
factor_frame.hist()
plt.show()

#---------------------------------------------------------------------------------------------------------------------#
# 2、取该300个股票2017-05-01到2017-06-01的因子值
# a、取沪深300成分股
hs300_frame = DataAPI.IdxConsGet(secID=u"",ticker=u"000300",isNew=u"",intoDate=u"",field=u"",pandas="1")
#DataAPI.IdxConsGet获取国内外指数的成分构成情况，包括指数成分股名称、成分股代码、入选日期、剔除日期等
hs300_frame = hs300_frame[hs300_frame.isNew==1]
# list格式的300个股票
tickers = list(hs300_frame.consTickerSymbol.values)

# b、取交易日历
calendar_frame = DataAPI.TradeCalGet(exchangeCD=u"XSHG,XSHE",beginDate=u"20170501",endDate=u"20170601",field=u"",pandas="1")
#DataAPI.TradeCalGet交易所交易日历      记录了上海证券交易所,深圳证券交易所,中国银行间市场,大连商品交易所,郑州商品交易所,上海期货交易所,中国金融期货交易所和香港交易所等交易所在日历日期当天是否开市的信息，其中上证、深证记录了自成立以来的全部日期是否开始信息。各交易日节假日安排通知发布当天即更新数据。

# 只取有交易的日期
calendar_frame = calendar_frame[calendar_frame.isOpen==1]


# c、遍历交易每一天，取因子数据
factor_frame_list = []
for tdate in calendar_frame.calendarDate.values:
    factor_frame = DataAPI.MktStockFactorsOneDayGet(tradeDate=tdate,ticker=tickers,field=["ticker",'tradeDate','ROE', 'DAVOL10'],pandas="1")
    #DAVOL10	10日平均换手率与120日平均换手率（Turnover Rate）之比。属于情绪类因子。数据更新时间：T日17点00分。
    factor_frame_list.append(factor_frame)#把每一天所有股票的因子数据的dataframe放入list

tfactor_frame = pd.concat(factor_frame_list)# 所有天的因子值合并到一个dataframe
# d、查看因子数据，每一个交易日所有股票的'DAVOL10'均值
print tfactor_frame.groupby(['tradeDate'])['DAVOL10'].mean().head()
ax = tfactor_frame.groupby(['tradeDate'])['DAVOL10'].mean().plot()
plt.show()
```
