```python
# coding:utf-8
# ===============================================================
# 名称：YF-工具-深度下跌底部抢反弹
# 功能：超跌反弹选股简化模型，寻找短期过度下跌有抢反弹机会的股票
# 创建时间：2017-12-11
# 最新更新时间：2018-02-02
# 策略概述：算法是给定时间区间、下跌幅度
# 和底部范围，找出A股市场内同时满足以下二个条件的个股以供进一步分析。
#     条件一：股价超跌：最小值在最大值之后并且（最大值-最小值）/最大值>=给定的下跌幅度；
#     条件二：当前股价处于底部范围：当前股价<=最小值*底部范围阀值
# 策略优缺点：
#     优点：
#       1）采用前复权数据，下跌幅度计算更真实；
#       2）以换手率来表现成交量情况可以看出量价之间的配合关系；
#       3）可以迅速给出A股市场所有符合条件的备选股
#     缺点：
#       1) 观察区间内上市的次新股会被忽略；
#       2）未考虑这些股票的基本面和消息面情况，需人工加以分析辨别；
#       3）仅用于选股，不能直接进行交易。
# 更新日志：
# 1）V0.1版实现了价格变化的分析
# 2）V0.2版采用前复权数据更真实反映价格变化，改进算法缩短运算时间
# 3）V0.3版增加换手率因子分析，量价配合更准确反映技术面的情况
# 4）V0.4版增加了数据导出功能
# ================================================================

# 导入模块
import numpy as np
import pandas as pd
import CAL.PyCAL as cal

# 开启缓存，当前Notebook所有DataAPI数据都会在内存进行缓存
DataAPI.settings.cache_enabled = True #调试时设为False
pd.set_option('display.max_rows',100)

# 设定条件
SPAN=cal.Period('12M') #默认设定时间区间为半年，如3个月3M，1年12M
LONG=30  #用于计算长期换手率均值
SHORT=5  #用于计算短期换手率均值

DOWN_RATE=0.5 #默认下跌幅度50%
BOTTOM_RANGE=0.1  #距离最低点不超过10%视为底部

print('*********************超跌反弹选股模型**********************')
print('一、策略思路：')
print(' 1、股价跌幅：观测期内股价（前复权）下跌幅度满足设定条件且当前股价仍在底部区域；')
print(' 2、成交量萎缩：短期成交量明显小于长期成交量')
print('设定条件：（可更改）')
print('观测期：12个月；股价下跌幅度：50%；底部范围：距最低点小于10%；换手率均值：长期30天，短期5天。')
print('结果说明：')
print(' 1、下跌幅度：（最高价-最低价）/最高价')
print(' 2、短期/长期成交量：短期成交量/长期成交量，>=1表示成交量增加，<1表示成交量萎缩')

# 调用CAL的日期处理功能，设定时间参数
today=cal.Date.todaysDate() #今日日期
calendar=cal.Calendar('China.SSE')  #建立A股市场日历对象
end=calendar.adjustDate(today, cal.BizDayConvention.Preceding) #时间区间结束日期
begin=calendar.adjustDate(today-SPAN, cal.BizDayConvention.Preceding) #时间区间开始日期
universe=set_universe('A') #获取A股股票代码

# ======================数据提取======================================
#获取股票代码和简称
df0=DataAPI.EquGet(equTypeCD=u"A",secID=universe,listStatusCD=u"",field=u"ticker,secShortName",pandas="1")

#获取行情数据
df=DataAPI.MktEqudGet(secID=universe,beginDate=begin,endDate=end,isOpen="",
                      field=u"ticker,tradeDate,closePrice,accumAdjFactor,turnoverRate",pandas="1")


#======================= 数据整理=====================================
df['adjPrice']=df.closePrice*df.accumAdjFactor  #计算前复权收盘价

df_cp=df[['ticker','tradeDate','adjPrice']] #提取前复权收盘价数据
df_cp.columns=['ticker','tradeDate','closePrice']  
df_tr=df[['ticker','tradeDate','turnoverRate']] #提取换手率数据

# 采用分组方法提取观察期内的价格最大值和最小值及其交易日期，最新价
cp_grouped=df_cp.groupby('ticker')

# 提取现价、最高价、最低价的函数
def get_hlc(group): 
    curr=group[-1:]
    tmp=group.sort_values('closePrice')
    low=tmp[:1]
    high=tmp[-1:]
    tmp=pd.merge(high,low,how='inner',on='ticker')
    tmp=pd.merge(tmp,curr,how='inner',on='ticker')
    return tmp
# 调用函数并整理标题、索引
df_cp_tmp=cp_grouped.apply(get_hlc)

# 整理列名
df_cp_tmp.columns=['ticker','highDate','highPrice','lowDate','lowPrice','currDate','currPrice']
# 判断股价跌幅及当前股价是否在底部范围
df_cp_tmp=df_cp_tmp[df_cp_tmp.highDate<df_cp_tmp.lowDate]  
df_cp_tmp['cp_chg']=(df_cp_tmp.highPrice-df_cp_tmp.lowPrice)/df_cp_tmp.highPrice
df_cp_tmp=df_cp_tmp[(df_cp_tmp.currPrice<=df_cp_tmp.lowPrice*(1+BOTTOM_RANGE))&(df_cp_tmp.cp_chg>=DOWN_RATE)] #价格条件：跌幅达到阀值&股价仍在底部

ticker_list=df_cp_tmp.ticker.tolist()
df_tr=df_tr[df_tr.ticker.isin(ticker_list)] 
tr_grouped=df_tr.groupby('ticker')

#计算换手率数据
def get_tr(group,long=30,short=5):
    tmp=group.iloc[0]
    tmp['long_tr']=group.iloc[-long:].turnoverRate.mean()
    tmp['short_tr']=group.iloc[-short:].turnoverRate.mean()
    return tmp

df_tr_tmp=tr_grouped.apply(get_tr,LONG,SHORT)
df_tr_tmp['tr_chg']=df_tr_tmp.short_tr/df_tr_tmp.long_tr
df_tr_tmp=df_tr_tmp.drop(['tradeDate','turnoverRate','long_tr','short_tr'],axis=1)
df_tr_tmp.sort_values('tr_chg')

# ==================数据合并&结果显示========================================
df_tmp=pd.merge(df0,df_cp_tmp,how='inner',on='ticker')
df_tmp=pd.merge(df_tmp,df_tr_tmp,how='inner',on='ticker')
df_tmp=df_tmp.sort_values('cp_chg',ascending=False).reset_index(drop=True) #排序并删除索引
df_tmp.columns=['股票代码','股票名称','高点日期','高点价格','低点日期','低点价格','当前时间','当前价格','下跌幅度','短期/长期成交量']
df_tmp.to_csv("CDFT"+today.strftime('%Y%m%d')+'.csv',encode='GBK') #数据导出到文件
df_tmp 
```
