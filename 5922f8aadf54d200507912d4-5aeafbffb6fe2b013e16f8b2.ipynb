# 多资产多策略策略平台

进入2017年，优矿向专业投资机构和量化爱好者正式推出支持多种资产的全新策略平台 **Quartz3**，该版本的策略平台支持以策略 API 规范（Strategy API Specification V3）编写的策略进行回测分析，模拟交易和实盘交易。

本文档描述 **Quartz3** 的策略写法和 API 定义，并给出了大量的简要示例。

点击此处，查看详细的[策略 API 规范](https://uqer.datayes.com/help/strategyManualV3/)。

## Quartz3为什么是更好的选择？

* **Quartz3** 支持标准级的**策略编写规范**，删减了大量冗余的策略 API ，让您只需要记忆少量的函数名，即可快速实现策略想法。
* **Quartz3** 扩充了策略交易的**资产范围**，从原来只支持股票、基金两种资产类型，扩充到支持**股票、场内外基金、指数、期货**四种资产的全资产平台。也就是说，一个策略中，既可以交易股票，又可以交易基金，又可以交易期货，实现真正意义上的**多空对冲策略**。在不久的将来，还会支持期权、债券、港股等资产类型。
* **Quartz3** 体现了策略**信号生成和交易执行相分离**的思想，可以灵活地进行各种资产的数据分析和交易撮合。
* **Quartz3** 支持**更多的策略特型**，如多因子裸多策略，市场中性策略，FOF策略，CTA期现策略等。
* **Quartz3** 会得到更充分的研究支持，新的示例会逐步扩充在**优矿知识库**中，辅助您进行更好的量化研究。
* **Quartz3** 在性能方面进行了诸多优化，您可以以**更快的运行速度**进行策略研究。

## 多因子选股策略（Long Only）
----
#### 下面的例子展示如何使用优矿进行多因子

- 回测区间：2015年1月1日~2016年1月1日
- 股票池：沪深300成分股
- 基准：沪深300
- 每周最后一个交易日换仓

#### 因子数据进行如下处理
- 因子选取：净利润增长率、权益收益率、RSI
- 因子处理：用到了去极值（winsorize）、标准化（standardize）、中性化（neutralize）处理
- 组合构建：等权配置


上述代码中出现了两个新的用法：

* `context.get_universe()`：

    表示交易日当天证券池中所有资产列表，与回测初始化配置中的 universe 对应。

    `exclude_halt` 参数用于调整是否去除停牌的股票

    详细帮助见 [context.get_universe()](https://uqer.datayes.com/help/strategyManualV3/#get_universe) 。

* `context.get_account('stock_account')`：

	表示关联的交易账户，用于下单、查询持仓、查询订单信息等。

* `account.order(stock, 100)`：

    用来下达买卖指令，表示买入100股 stock 股票。如果amount为正数，表示买入；为负数，则表示卖出。

    详细帮助见 [account.order()](https://uqer.datayes.com/help/strategyManualV3/#order) 。

## 期货日内策略
----
### 下面的例子展示如何使用优矿进行期货日内策略

- 回测区间：2015年11月1日~2016年12月28日
- 股票池：螺纹钢合约
- 基准：沪深300
- 每天运行一次

### 调仓逻辑
- 计算K线图的支撑位和压力位，随时观察实时价格
- 如果价格超过压力位，则平空做多
- 如果价格下穿支撑位，则平多做空
- 策略只做一天，每天收盘，全部平仓

## 基金轮动策略
----
### 下面的例子展示如何使用优矿进行基金轮动策略

- 回测区间：2015年1月1日~2017年1月1日
- 股票池：南方中证500ETF、易方达沪深300ETF
- 基准：沪深300
- 每天运行策略算法

### 调仓逻辑
- 比较大盘基金和小盘基金的最近20天收益率。
- 如果大盘的收益率超过小盘的收益率，则持有大盘。
- 如果小盘的收益率超过大盘的收益率，则持有小盘。
- 如果大小盘收益率均为负，则清仓。
