# 封板率统计

```python
import pandas as pd
from pandas import DataFrame,Series
from CAL.PyCAL import *
import time
import matplotlib.pyplot as plt
pd.set_option('display.width', 200)


def get_rate(today):
    # today = '2015-12-24'
    pday = today.strftime('%Y-%m-%d')
    today = today.strftime('%Y%m%d')
    yesterday = today
    allSecList = DataAPI.EquGet(equTypeCD = 'A', field = ['secID'])['secID'].tolist()
    h_10=0   ##最高触及涨停
    c_10=0   ##收盘为涨停
    for i in range(0, len(allSecList), 200):
        sub = allSecList[i:min(len(allSecList), i+200)]
        df_info = DataAPI.MktEqudGet(secID = sub, beginDate = yesterday, endDate = yesterday, field = 'secID,closePrice,preClosePrice,highestPrice,lowestPrice')
        if len(df_info)<5:
            # print 'return'
            return 0,0,0,0
        df_info['up_target']=(df_info['preClosePrice']*1.1).round(2)
        df_info['dn_target']=(df_info['preClosePrice']*0.9).round(2)
        high_10 = df_info[df_info.highestPrice == df_info.up_target]         ##触及涨停数
        close_10 = df_info[(df_info.highestPrice == df_info.up_target) & (df_info.highestPrice == df_info.closePrice)]         ##最终涨停数
        h_10+=len(high_10)
        c_10+=len(close_10)
    keep_rate = round(float(c_10)/float(h_10),2)
    return c_10,h_10,keep_rate,pday

rateframe=DataFrame()
index = 0
cal = Calendar('China.SSE')
startDay = '2015-11-20'
endDay = '2016-01-11'
Day = cal.advanceDate(startDay, '+1B',  BizDayConvention.Following)
while(Day.strftime('%Y-%m-%d') <= endDay):
    (c_10,h_10,keep_rate,date) = get_rate(Day)
    if h_10>1:
        rateframe.loc[index,'date']=date
        rateframe.loc[index,'nhigh']=h_10
        rateframe.loc[index,'nclose']=c_10
        rateframe.loc[index,'keep_rate']=keep_rate
        index +=1
    Day = cal.advanceDate(Day.strftime('%Y-%m-%d'), '+1B',  BizDayConvention.Following)
    

pd.set_option('display.width', 200)
fig=plt.figure(figsize=[20,8])

ax1=fig.add_subplot(111)
time_slots=list(rateframe.index.values)
time_slot=[]
for i in range(0,len(time_slots)):
    if i%5==0:
        time_slot.append(time_slots[i])
time_str=rateframe.date.values
line11,=ax1.plot(rateframe.index.values,rateframe.keep_rate.values,'r-')
line12,=ax1.plot(rateframe.index.values,rateframe.keep_rate.values,'ro')
ax1.set_xticks(time_slot)
ax1.set_xticklabels(time_str[time_slot])
ax1.grid(True)
# plt.title("封板率")
plt.show()
```
