# 可转债双低轮动策略

```python
# 获取某日可转债指数所有成分
def get_bond(tradeDate):
    bond = DataAPI.IdxConsGet(ticker=u"000832",intoDate=tradeDate,field=u"consID",pandas="1")
    consID = bond['consID']
    return list(consID.values)
def getCurrentResult(lastDate, currentDate, result, to_hold):
    #获取本轮收益
    if len(to_hold) > 0:
        price = DataAPI.MktBonddGet(secID=to_hold,beginDate=lastDate,endDate=currentDate,\
                                  field=u"secID,tradeDate,openPrice,closePrice",pandas="1")
        price = price.groupby(['secID','tradeDate']).mean().unstack().T
        openPrice = price.loc['openPrice'].iloc[0]
        closePrice = price.loc['closePrice'].iloc[-2]#调仓日前一天卖出
        #print openPrice,closePrice
        mean = (closePrice/openPrice-1).mean()
        if 3 > mean > -1:
            mean = mean
        else:
            mean = 0
        result[currentDate] = mean * len(to_hold) / maxHold   
    else:
        result[currentDate] = 0
```

# 以下是模拟策略

```python
#获取每周最后一个交易日
cal=DataAPI.TradeCalGet(exchangeCD=u"XSHG",beginDate=u"20130101",endDate=u"20220701",isOpen=u"1",field=u"",pandas="1")
cal = cal.loc[cal['isWeekEnd']==1]
cal.index = range(len(cal))
print(len(cal))

result = {}
to_hold = []
maxHold = 10#最大持仓，每个转债最多10%
for i in range(len(cal)):
    currentDate = cal.loc[i]['calendarDate']
    if i > 0:
        lastDate = cal.loc[i-1]['calendarDate']
        getCurrentResult(lastDate, currentDate, result, to_hold)
    else:
        result[currentDate] = 0
    bond = get_bond(currentDate)#获取所有可转债
    #获取可转债溢价
    bondPremRatio = DataAPI.MktConsBondPerfGet(beginDate=currentDate,endDate=currentDate,secID=bond,
                       field=u"secID,bondPremRatio",pandas="1")

    bondPremRatio.set_index('secID',inplace=True)
    bondPremRatio.columns=['value']
    #排除当日停牌的
    price = DataAPI.MktBonddGet(secID=bond,beginDate=currentDate,endDate=currentDate,
                                  field=u"secID,openPrice",pandas="1")
    price = price[price.openPrice>0]
    price = price[price.openPrice<150]
    price.set_index('secID',inplace=True)
    price.columns=['value']
    z = price + bondPremRatio
    z.dropna(inplace=True)
    z.sort_values(by='value', ascending=True, inplace=True)
    to_hold = list(z.iloc[:maxHold].index)
    print(currentDate, len(to_hold), to_hold)
print('end')
```
