```python
#设定股票池、日期以及日历
universe = DynamicUniverse('HS300')
begin_date = '20100101'
end_date = '20180228'
trade_date_list = DataAPI.TradeCalGet(exchangeCD=u"XSHG",beginDate=begin_date,endDate=end_date,\
                                      isOpen=u"1",field=u"calendarDate,isOpen,isMonthEnd",pandas="1")
trade_date_list = trade_date_list[trade_date_list['isMonthEnd']==1]
trade_date_list=trade_date_list['calendarDate'].tolist()
len(trade_date_list)
```

```python
#获取沪深300多日的因子值
factor = pd.DataFrame()
for trade_date in trade_date_list:
    current_universe = universe.preview(date =trade_date)
    current_factor = DataAPI.MktStockFactorsOneDayGet(tradeDate=trade_date,secID=current_universe,\
                                                     field=u"secID,tradeDate,PE,PB,PS,ROE,PCF",pandas="1")
    factor= factor.append(current_factor)   
factor.head()
```

```python
pe = factor.pivot(index ='tradeDate',columns='secID',values='PE')
pb = factor.pivot(index ='tradeDate',columns='secID',values='PB')
ps = factor.pivot(index ='tradeDate',columns='secID',values='PS')
roe = factor.pivot(index ='tradeDate',columns='secID',values='ROE')
pcf = factor.pivot(index ='tradeDate',columns='secID',values='PCF')
factor_dict = {'pe':pe,'pb':pb,'ps':ps,'roa':roa,'pcf':pcf}
```

```python
#获取沪深300一定期间的月度收益率
month_return = DataAPI.MktEqumAdjGet(beginDate=u"20100101",endDate=u"20180321",isOpen=u"",\
                                     field=u"endDate,secID,chgPct",pandas="1")
month_return = month_return.pivot(index ='endDate',columns ='secID',values='chgPct')
forward_1m_ret =month_return.shift(-1).loc[:,factor.secID.unique()]
forward_1m_ret.head()
```

```python
#等分权重因子
factor = (pe+pb+ps+roe+pcf)/5
factor.head()
```

```python
#非等分权重因子
factor = 0.2*pe+0.2*pb+0.2*ps+0.3*roe+0.1*pcf
factor.head()
#factor.loc['2010-01-29']
```

```python
#定义返回相关系数函数
def get_rank_ic(factor,forward_return):
    commom_index = factor.index.intersection(forward_return.index)
    ic_data = pd.DataFrame(index=commom_index,columns = ['IC','P_values'])
    for dt in ic_data.index:
        cor = pd.DataFrame(factor.loc[dt])
        ret = pd.DataFrame(forward_return.loc[dt])
        cor.columns = ['factor']
        ret.columns = ['ret']
        cor['ret']=ret['ret']
        #cor = cor[~pd.isnull(cor['factor'])][~pd.isnull(cor['ret'])]
        cor = cor.dropna()
        if len(cor)<5:
            continue
        ic,p_value = st.spearmanr(cor['factor'],cor['ret'])
        ic_data.loc[dt,'IC']=ic
        ic_data.loc[dt,'P_values'] = p_value
    return ic_data    
```

```python
pe_ic = get_rank_ic(pe,forward_1m_ret)
pe_ic.head()
```

```python
pb_ic = get_rank_ic(pb,forward_1m_ret)
ps_ic = get_rank_ic(ps,forward_1m_ret)
roe_ic = get_rank_ic(roe,forward_1m_ret)
pcf_ic = get_rank_ic(pcf,forward_1m_ret)
```

```python
#滚动12个月取平均
pe_rolling_ic = pe_ic.rolling(12).mean()
pb_rolling_ic = pb_ic.rolling(12).mean()
ps_rolling_ic = ps_ic.rolling(12).mean()
roe_rolling_ic = roe_ic.rolling(12).mean()
pcf_rolling_ic = pcf_ic.rolling(12).mean()
```

```python
pcf_rolling_ic.head(13)
```

```python
#将五个因子组合为DF
all_rolling_ic_df = pd.concat((pe_rolling_ic['IC'],pb_rolling_ic['IC'],ps_rolling_ic['IC'],roe_rolling_ic['IC'],pcf_rolling_ic['IC']),axis= 1)
all_rolling_ic_df.head(13)
```

```python
factor_weight = all_rolling_ic_df.divide(all_rolling_ic_df.abs().sum(axis = 1),axis = 0)
factor_weight.columns = ['pe','pb','ps','roe','pcf']
factor_weight.head(13)
```

```python
#计算各个因子的权重
factor = pe.multiply(factor_weight['pe'],axis = 0)+pb.multiply(factor_weight['pb'],axis = 0)+ps.multiply(factor_weight['ps'],axis = 0)+roe.multiply(factor_weight['roe'],axis = 0)+pcf.multiply(factor_weight['pcf'],axis = 0)
factor.head(13)
```
