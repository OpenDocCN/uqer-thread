# `思路整理`
***

- #### 筛选对于市场波动有一定预测能力的指标，备选：
    - #### 实体经济指标：`PMI(采购经理人指数)`
    - #### 物价水平指标：`CPI(消费者价格指数)`、`PPI(生产者价格指数)`
    - #### 货币财政指标：`M1(狭义货币量)`、`M2(广义货币量)`


- #### 选取样本期间进行模型训练，得出市场回报预测模型；

- #### 在之后的期间利用所得模型预测市场收益，大于一定门槛值则持有沪深300ETF，否则空仓。



# `所用模型`
***

\[R_t = \beta_0 +\beta_1 PMI_{t-1}+\beta_2 CPI_{t-2}+ \beta_3PPI_{t-2}+\beta_4M1MOY_{t-2}+\beta_5M2MOY_{t-2}\]
- #### 参数从左至右分别为：$t$月股指收益率、$t-1$月PMI、$t-2$月CPI(增长率)、$t-2$月PPI(增长率)、$t-2$月M1同比增长率、$t-2$月M2同比增长率；

- #### 滞后期选择是根据数据的获得时间，PMI数据在每月月初公布，而其他宏观数据在每月月中公布。如在8月1日希望预测8月股指回报，能够获取的数据是7月的PMI数据和8月的CPI、PPI、M1、M2.


# `实战演练！`
***

- #### 引入工具包并定义全局设定：

```python
# libs
from __future__ import division
import seaborn as sns
import datetime as dt
import pandas as pd
import matplotlib.pyplot as plt
import math
import string
from sklearn import linear_model
from CAL.PyCAL import *

# settings
sns.set_style('whitegrid')
fsize = (12, 12*0.618)
today = dt.datetime.today().strftime('%Y%m%d')
DataAPI.settings.cache_enabled = True
```

```python
def monthBefore(year = 2016, month = 1, delta = 2):
    # 获取delta个月之前的年月
    rtnMonth = (month + 12 - (delta%12)) % 12
    rtnMonth = 12 if rtnMonth == 0 else rtnMonth
    yearDelta = int((delta - month) / 12) + 1
    rtnYear = year - yearDelta if delta >= month else year
    return dt.datetime(rtnYear, rtnMonth, 1).strftime('%Y-%m')
```

- #### 选一只上市时间长的ETF测试择时效果：

```python
DataAPI.SecIDGet(ticker = '160706')[1:2]
```

- #### 用到的宏观数据指标是：

```python
DataAPI.ChinaMacroInfoGet(indicID = ['M030000157', 'M030000012', 'M100000004', 'M100000005', 'M020000008', 'M040000013', 'M100000007', 'M100000008'])
```

- #### 获取每月指数行情数据：

```python
cal = DataAPI.TradeCalGet(exchangeCD=u"XSHE",beginDate= '20050701',endDate = today , field= 'calendarDate,isMonthEnd')
cal = cal[cal.isMonthEnd == 1]
cal['stockIndex'] = map(lambda x: DataAPI.MktIdxdGet(tradeDate= x, ticker = '000001', 
                                                     field = 'closeIndex').values[0][0], cal.calendarDate.values)
cal.calendarDate = cal.calendarDate.apply(lambda x: x[:7])
cal = cal.set_index('calendarDate').drop('isMonthEnd', axis = 1)
```

- #### 数据组合与处理：

```python
macroData = DataAPI.ChinaMacroDataGet(indicID = 
            ['M030000157', 'M030000012', 'M100000004', 'M100000005', 'M020000008', 'M040000013', 'M100000007', 'M100000008'], 
            beginDate= '20050701').pivot(index = 'periodDate', values = 'dataValue', columns = 'indicName')
macroData.columns = ['PMI', 'CPI_MOM', 'PPI_MOM', 'VAI_SS_MOM', 'M1', 'M2', 'M1_rate', 'M2_rate']
macroData['CPI'] = 1 + macroData.CPI_MOM / 100
macroData['CPI'] = macroData.CPI.cumprod()
macroData['PPI'] = 1 + macroData.PPI_MOM / 100
macroData['PPI'] = macroData.PPI.cumprod()
macroData['M1_index'] = macroData.M1 / macroData.M1[:1][0]
macroData['M2_index'] = macroData.M2 / macroData.M2[:1][0]
macroData['M2mM1_index'] = macroData.M2 - macroData.M1
macroData['M2mM1_index'] = macroData['M2mM1_index'] / macroData['M2mM1_index'][:1][0]

macroData.index = map(lambda x: x[:7], macroData.index)
tmData = macroData.merge(cal, left_index = True, right_index = True)
tmData.stockIndex = tmData.stockIndex / tmData.stockIndex[:1][0]
tmData['indexReturn'] = tmData.stockIndex.pct_change()

tmData.tail(5)
```

- #### 构建模型训练数据集：
    - #### 将数据调整到相应滞后期；
    - #### 选取训练期间 —— 这里用`2005.07 - 2014.10`作为训练期间。 

```python
toDateRegData = pd.DataFrame()
toDateRegData['CPI_MOM'] = tmData.CPI_MOM[:-2]
toDateRegData['PPI_MOM'] = tmData.PPI_MOM[:-2]
toDateRegData['M1_rate'] = tmData.M1_rate[:-2]
toDateRegData['M2_rate'] = tmData.M2_rate[:-2]
toDateRegData['PMI'] = tmData.PMI[1:-1].values
toDateRegData['indexReturn'] = tmData.indexReturn[2:].values
regData = toDateRegData[:'2014-09']
regData.tail()
```

- #### 拟合线性回归模型：
    - #### 模型能够解释16%的股指回报率变化情况；
    - #### 模型预测股指涨跌的准确率是61.3%.

```python
regr = linear_model.LinearRegression()
regr.fit(regData[['CPI_MOM', 'PPI_MOM', 'M1_rate', 'M2_rate', 'PMI']], regData.indexReturn)
print "R-square of this regression is %f."%regr.score(regData[['CPI_MOM', 'PPI_MOM', 'M1_rate', 'M2_rate', 'PMI']], regData.indexReturn)
regData['predictReturn'] = regr.predict(regData[['CPI_MOM', 'PPI_MOM', 'M1_rate', 'M2_rate', 'PMI']])
sns.pairplot(regData[['indexReturn', 'predictReturn']], x_vars= 'predictReturn', y_vars= 'indexReturn', kind = 'reg', size= 7)
print "符号准确率是%f."%(len(regData[regData.predictReturn*regData.indexReturn > 0]) / len(regData))
```

- #### 对于模型训练外期间（这里从2014.11开始）进行回测，如果月初预测股指回报率高于门槛值，全仓持有沪深300ETF；否则空仓。验证择时效果：
    - #### 能够较为准确地分辨上涨和下跌期间；
    - #### 回撤小，回报高。

# `结论与优化`
***

- #### 本篇给出了宏观因子择时的一种实现方式：采用宏观因子滞后期数据（避免预测时使用未来数据）与历史股指回报拟合模型，在训练样本外期间使用得到参数预测股指回报，并进行相应操作。样本外回测表现不错，回撤小，回报高。

- #### 这里只提供一种思路，宏观因子选取、拟合模型等都还有不同的选择。
