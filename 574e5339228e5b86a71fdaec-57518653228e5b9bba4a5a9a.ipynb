可以看到，在 `10:42:10` 秒的时候，突然出现了一个将近400手的卖单，把价格拉低将近200点。下面我们来仔细看看数据，回顾当时究竟发生了什么？

```python
df2 = df[(df.dataTime < '10:42:30') & (df.dataTime > '10:42:05')]
df2[['dataTime', 'lastPrice', 'change', 'changePCT', 'Volume', 'bidPrice1', 'askPrice1', 'addInterest', 'buyVol', 'sellVol']]
```

`原因推断`

- 10:42:11 时，有一个 398 张的大单，把点数拉低 146.8 点，但同时增仓量只有170手
- 在这个 398 张大单的导火索下，在短短一秒钟内触发了 248 手的大单，瞬间打到跌停价，但注意同时增仓量却只有 44 张；
- 紧接着又在跌停价成交了 7 秒钟，祝福在这个价位开多的兄弟们，是时候该发红包了；至于在这10秒钟平仓的兄弟们，坚持住，这只是人生的一个小坎坎；
