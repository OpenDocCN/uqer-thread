```python
2.unstack函数
速度远快于循环和groupby方式，运算更加自由
参考资料:http://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.unstack.html
```

```python
import pandas as pd
df = DataAPI.MktEqudGet(ticker=u"000001,000002,000004",beginDate=u"20180502",endDate=u"20180507",isOpen="",field=u"ticker,tradeDate,closePrice,openPrice",pandas="1")
print (df)
df1 = df[["ticker","tradeDate","closePrice"]].set_index(['tradeDate', 'ticker'])#设置两重索引
print (df1)
df2 = df1.unstack()#对两重索引重排变成三维结构
df3 = df.set_index(['tradeDate', 'ticker']).unstack()
print(df2)
print (df3)
print('=======================计算MA5=======================')
print(df2.rolling(5).mean())
print(df2.rolling(window=5,min_periods=1).mean())#设置min_periods=1，当不足window数量也会求均值
print('=====================计算每日涨跌=====================')
print(df2.pct_change())#此函数将每个元素与其前一个元素进行比较，并计算变化百分比
print('====================计算每日最高股价====================')
print(df2.max(axis=1))#某一天的最高股价
```

```python
df4 = df2.copy(deep=True)
df5 = df4.stack()#有两层索引
df6 = df4.stack().reset_index()#重置索引
print(df5)
print(df6)
```

```python
#unstack方法与groupby方法比较

import time
#获取行情数据
df = DataAPI.MktEqudAdjGet(secID=u"",ticker=u"",beginDate=u"20180101",endDate=u"20180509",field=u"ticker,tradeDate,closePrice",pandas="1")
print(df.head())
print('='*40)
df = df.sort_values(['ticker', 'tradeDate'])#按'ticker', 'tradeDate'分别升序排列

df_groupby = df.copy(deep=True)

#用unstack方法计算
t1 = time.time()
#df1 = df.set_index(['tradeDate', 'ticker']).unstack()#第一个索引变为行索引，第二个索引变为列索引,三维数据结构，数据从长格式’转换为‘宽格式’ ，构建层次化索引
#df1 = df.set_index(['tradeDate', 'ticker']).unstack().pct_change()#求变化百分比
#df1 = df.set_index(['tradeDate', 'ticker']).unstack().pct_change().stack(dropna=False)#默认参数dropna=True
#print (df1.head())

df = df.set_index(['tradeDate', 'ticker']).unstack().pct_change().stack(dropna=False).reset_index()#先索引重设，进行结构重排（构建层次化索引），再做相应计算，最后结构还原，重置index（层次化索引重置）
df = df.sort_values(['ticker', 'tradeDate'])
print(df.head())
t2 = time.time()
print('unstack: ', t2 - t1)

print('='*40)
print('='*40)
#用groupby方法计算
t3 = time.time()
df_groupby['closePrice'] = df_groupby.groupby('ticker')['closePrice'].pct_change()#先groupby在进行列操作，而unstack之后是直接列操作，速度更快
df_groupby = df_groupby.sort_values(['ticker', 'tradeDate'])
print(df_groupby.head())
t4 = time.time()
print('groupby: ', t4 - t3)
```

3.pd.Series.first_valid_index、pd.Series.last_valid_index
参考资料:http://pandas.pydata.org/pandas-docs/stable/generated/pandas.Series.first_valid_index.html

```python
df = DataAPI.MktEqudGet(ticker=u"000001,000002,000004,000008",beginDate=u"20180502",endDate=u"20180509",isOpen="",field=u"ticker,tradeDate,closePrice",pandas="1")
df = df.set_index(['tradeDate', 'ticker'])['closePrice'].unstack()
df.iloc[0, 0] = np.nan
df.iloc[0, 1] = np.nan
df.iloc[1, 1] = np.nan
df.iloc[5, 0] = np.nan
print(df)
print('='*40)
print(df.apply(pd.Series.first_valid_index, axis=0))#第一个非Nan数据出现的位置
print(df.apply(pd.Series.last_valid_index, axis=0))#最后一个非Nan数据出现的位置
```
