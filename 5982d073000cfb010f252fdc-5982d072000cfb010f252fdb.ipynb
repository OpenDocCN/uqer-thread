```python
import matplotlib.pyplot as plt
from matplotlib import rc
rc('mathtext', default='regular')
import seaborn as sns
sns.set_style('white')
from matplotlib import dates
import datetime
import numpy as np
import pandas as pd
import time
import scipy.stats as st
from CAL.PyCAL import *    # CAL.PyCAL中包含fon

# DataAPI.FdmtISQGet(ticker=u"000002",secID="",endDate="",beginDate="",beginYear=u"",endYear=u"",reportType=u"",field=u"",pandas="1")  利润表
# DataAPI.FdmtCFQGet(ticker=u"000002",secID="",endDate="",beginDate="",beginYear=u"",endYear=u"",reportType=u"",field=u"",pandas="1")  现金流量表

universe = set_universe("A")
beginDate = '2010-01-01'
endDate = '2016-12-31'

def getFinancialData(universe,beginDate,endDate):
    ratio = pd.DataFrame()
    for stock in universe:
        IS_Data = DataAPI.FdmtISQGet(secID=stock,endDate=endDate,beginDate=beginDate,field=u"endDate,NIncome",pandas="1")
        IS_Data = IS_Data.sort('endDate',ascending=True)
        CF_Data = DataAPI.FdmtCFQGet(secID=stock,endDate=endDate,beginDate=beginDate,field=u"endDate,NCFOperateA",pandas="1")
        CF_Data = CF_Data.sort('endDate',ascending=True)
        CF_Data[stock] = CF_Data['NCFOperateA'] / IS_Data['NIncome']
        
        tmp_ratio = CF_Data[['endDate',stock]]
        
        if ratio.empty:
            ratio = tmp_ratio
        else:
            ratio = ratio.merge(tmp_ratio ,on='endDate',how='outer')
    ratio.set_index('endDate',inplace=True)
    ratio.index.name = 'tradeDate'
    ratio.to_csv('CFIS_Ratio.csv')

getFinancialData(universe,beginDate,endDate)

```

```python
cal_dates = DataAPI.TradeCalGet(exchangeCD=u"XSHG",beginDate=beginDate,endDate=endDate)
quarter_dates = cal_dates[cal_dates['isQuarterEnd'] == 1]['calendarDate']
month_dates = cal_dates[cal_dates['isMonthEnd'] == 1]['calendarDate']
data = pd.DataFrame()
window = 3
count = 0
for stock in universe:
    count += 1
    if count % 100 == 0:
        print '已经有'+str(count)+'家公司完成'
    tmp_data = DataAPI.MktEqumGet(secID=stock,beginDate=beginDate,endDate=endDate,field=u"endDate,closePrice",pandas="1")
    tmp_data[stock] = tmp_data['closePrice'].shift(-window) / tmp_data['closePrice'] - 1
    tmp_data = tmp_data[tmp_data['endDate'].isin(quarter_dates)]
    del(tmp_data['closePrice'])
    tmp_data.sort('endDate',inplace=True)
    
    if data.empty:
        data = tmp_data
    else:
        data = data.merge(tmp_data,on='endDate',how='outer')
data.set_index('endDate',inplace=True)
data.index.name = 'tradeDate'
data.to_csv('stock_Ret.csv')

```

```python
import matplotlib.pyplot as plt
from matplotlib import rc
rc('mathtext', default='regular')
import seaborn as sns
sns.set_style('white')
from matplotlib import dates
import datetime
import numpy as np
import pandas as pd
import time
import scipy.stats as st
from CAL.PyCAL import *    # CAL.PyCAL中包含fon

CFIS_Ratio = pd.read_excel('CFIS_Ratio.xlsx')
ret_Data = pd.read_excel('stock_Ret.xlsx')
CFIS_Ratio.set_index('tradeDate',inplace=True)
ret_Data.set_index('tradeDate',inplace=True)

ic_data = pd.DataFrame(index = CFIS_Ratio.index,columns=['IC','pValue'])
# IC值探析
for dt in CFIS_Ratio.index:
    tmp_CFIS = pd.DataFrame(CFIS_Ratio.ix[dt])
    tmp_Ret = pd.DataFrame(ret_Data.ix[dt])
    tmp_CFIS.columns = ['CFIS']
    tmp_Ret.columns = ['Ret']
    tmp_CFIS['Ret'] = tmp_Ret['Ret']
    tmp = tmp_CFIS[~np.isnan(tmp_CFIS['CFIS'])][~np.isnan(tmp_CFIS['Ret'])]
    if len(tmp) < 10:
        print '存在诸多空值，请返回'
        continue
    
    ic,pvalue = st.spearmanr(tmp['CFIS'],tmp['Ret'])
    ic_data['IC'][dt] = ic
    ic_data['pValue'][dt] = pvalue

print 'IC均值: ', ic_data['IC'].mean()
print 'IC中位数: ', ic_data['IC'].median()
print 'IC测试期数: ', len(ic_data),' 正相关性IC测试期数： ', len(ic_data[ic_data.IC>0]),' 负相关性IC测试期数： ', len(ic_data[ic_data.IC<0])
    
    
```

```python
fig = plt.figure(figsize=(12, 6))
ax1 = fig.add_subplot(111)

lns1 = ax1.plot(ic_data.index, ic_data.IC, label='IC')

lns = lns1
labs = [l.get_label() for l in lns]
ax1.legend(lns, labs, loc='', ncol=3, mode="", borderaxespad=0., fontsize=12)
ax1.set_ylabel(u'相关系数', fontproperties=font, fontsize=16)
ax1.set_xlabel(u'日期', fontproperties=font, fontsize=16)
ax1.set_title(u"CFIS和之后1个季度收益的秩相关系数", fontproperties=font, fontsize=16)
ax1.grid()
```

```python
# 测试十组分位
n_quantile = 10
cols = [i+1 for i in range(n_quantile)]

excess_return = pd.DataFrame(index = CFIS_Ratio.index,columns = cols)

for dt in excess_return.index:
    tmp_ratio = CFIS_Ratio.ix[dt].dropna()
    tmp_ret = ret_Data.ix[dt].dropna()
    tmp_MeanRet = ret_Data.ix[dt].mean()
    results = []
    for i in range(n_quantile):
        down = tmp_ratio.quantile(i*1.0/n_quantile)
        up = tmp_ratio.quantile((i+1)*1.0/n_quantile)
        tmp_index = tmp_ratio[(tmp_ratio >= down) & (tmp_ratio <= up)].index
        mean_tmp = tmp_ret[tmp_index].mean() - tmp_ret.mean()
        results.append(mean_tmp)
    excess_return.ix[dt] = results
excess_return.dropna(inplace=True)
excess_return
    
```

```python
fig = plt.figure(figsize=(12, 6))
ax1 = fig.add_subplot(111)

excess_returns_means_dist = excess_return.mean()
lns1 = ax1.plot(excess_returns_means_dist.index, excess_returns_means_dist.values, '--o', label='IC')
excess_dist_plus = excess_returns_means_dist[excess_returns_means_dist>0]
excess_dist_minus = excess_returns_means_dist[excess_returns_means_dist<0]
lns2 = ax1.bar(excess_dist_plus.index, excess_dist_plus.values, align='center', color='r', width=0.35)
lns3 = ax1.bar(excess_dist_minus.index, excess_dist_minus.values, align='center', color='g', width=0.35)

ax1.set_xlim(left=0.5, right=len(excess_returns_means_dist)+0.5)
ax1.set_ylim(-0.02, 0.02)
ax1.set_ylabel(u'超额收益', fontproperties=font, fontsize=16)
ax1.set_xlabel(u'十分位分组', fontproperties=font, fontsize=16)
ax1.set_xticks(excess_returns_means_dist.index)
ax1.set_xticklabels([int(x) for x in ax1.get_xticks()], fontproperties=font, fontsize=14)
ax1.set_yticklabels([str(x*100)+'0%' for x in ax1.get_yticks()], fontproperties=font, fontsize=14)
ax1.set_title(u"CFIS因子超额收益", fontproperties=font, fontsize=16)
ax1.grid()
```
