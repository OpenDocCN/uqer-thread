### 1.1 Figures和Subplots

+ Figure是画布，必须先生成画布
+ 之后再在Figure上生成Subplots，之后在Subplot上才能作图
+ Subplots可生成多个
+ 如果没有指定，图会画在最近提到的Subplot上

```python
fig = plt.figure()
ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)
from numpy.random import randn
plt.plot(randn(50).cumsum(), 'k--')
ax1.hist(randn(100), bins=20, color='k', alpha=0.3)
ax2.scatter(np.arange(30), np.arange(30) + 3 * randn(30))
```

调整Subplot周围的间距，使用函数`plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=None, hspace=None)`

下图将间距收缩到了0：

```python
fig, axes = plt.subplots(2, 2, sharex=True, sharey=True)
for i in range(2):
    for j in range(2):
        axes[i, j].hist(randn(500), bins=50, color='k', alpha=0.5)
plt.subplots_adjust(wspace=0, hspace=0)
```

### 1.2 颜色、标记和线型

使用color、linestyle、marker属性可以设定颜色、线型、标记：

```python
plt.figure()
plt.plot(randn(30).cumsum(), color='k', linestyle='dashed', marker='o')
```

也可以使用简略形式，具体颜色、线型、标价的代码可查看`plot()`函数的文档

```python
plt.figure()
plt.plot(randn(30).cumsum(), 'ko--')
```

在线型图中，非实际数据点默认是按线性方式插值的，可以通过drawstyle选项修改：

```python
data = randn(30).cumsum()
plt.plot(data, 'k--', label='Default')
plt.plot(data, 'k-', drawstyle='steps-post', label='steps-post')
plt.legend(loc='best')
```

### 1.3 刻度、标签和图例

设置标题、轴标签、刻度以及刻度标签

```python
fig = plt.figure(); ax = fig.add_subplot(1, 1, 1)
ax.plot(randn(1000).cumsum())

ticks = ax.set_xticks([0, 250, 500, 750, 1000])
labels = ax.set_xticklabels(['one', 'two', 'three', 'four', 'five'], rotation=30, fontsize='small')
ax.set_title('My first matplotlib plot')
ax.set_xlabel('Stages')
```

添加图例

```python
fig = plt.figure(); ax = fig.add_subplot(1, 1, 1)
ax.plot(randn(1000).cumsum(), 'k', label='one')
ax.plot(randn(1000).cumsum(), 'k--', label='two')
ax.plot(randn(1000).cumsum(), 'k.', label='three')

ax.legend(loc='best')
```

### 1.4 注解

通过`.annotate`方法添加：

```python
from datetime import datetime

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

data = pd.read_csv('spx.csv', index_col=0, parse_dates=True)
spx = data['SPX']

spx.plot(ax=ax, style='k-')

crisis_data = [
    (datetime(2007, 10, 11), 'Peak of bull market'),
    (datetime(2008, 3, 12), 'Bear Stearns Fails'),
    (datetime(2008, 9, 15), 'Lehman Bankruptcy')
]

for date, label in crisis_data:
    ax.annotate(label, xy=(date, spx.asof(date) + 50),
                xytext=(date, spx.asof(date) + 200),
                arrowprops=dict(facecolor='black'),
                horizontalalignment='left', verticalalignment='top')

# Zoom in on 2007-2010
ax.set_xlim(['1/1/2007', '1/1/2011'])
ax.set_ylim([600, 1800])

ax.set_title('Important dates in 2008-2009 financial crisis')
```

## 2 pandas中的绘图函数

### 2.1 线形图

对于Series和DstaFrame对象，可以使用`.plot()`方法直接生成线形图。

```python
s = Series(np.random.randn(10).cumsum(), index=np.arange(0, 100, 10))
s.plot()
```

对于DataFrame对象，会在同一个Subplot中对每一列做线形图，并自动添加图例：

```python
df = DataFrame(np.random.randn(10, 4).cumsum(0),
               columns=['A', 'B', 'C', 'D'],
               index=np.arange(0, 100, 10))
df.plot()
```

### 2.2 柱形图

在`.plot()`中使用kind='bar'或kind='barh'可绘制柱形图或条形图：

```python
fig, axes = plt.subplots(2, 1)
data = Series(np.random.rand(16), index=list('abcdefghijklmnop'))
data.plot(kind='bar', ax=axes[0], color='k', alpha=0.7)
data.plot(kind='barh', ax=axes[1], color='k', alpha=0.7)
```

对于DataFrame，则每一行的值分为一组：

```python
df = DataFrame(np.random.rand(6, 4),
               index=['one', 'two', 'three', 'four', 'five', 'six'],
               columns=pd.Index(['A', 'B', 'C', 'D'], name='Genus'))
df
df.plot(kind='bar')
```

可以使用stacked=True绘制堆积柱形图：

```python
df.plot(kind='barh', stacked=True, alpha=0.5)
```

### 2.3 直方图和密度图

```python
tips = pd.read_csv('tips.csv')
tips['tip_pct'] = tips['tip'] / tips['total_bill']
tips['tip_pct'].hist(bins=50)
```

```python
tips['tip_pct'].plot(kind='kde')
```

将密度图和密度图绘制在一起：

```python
comp1 = np.random.normal(0, 1, size=200)  # N(0, 1)
comp2 = np.random.normal(10, 2, size=200)  # N(10, 4)
values = Series(np.concatenate([comp1, comp2]))
values.hist(bins=100, alpha=0.3, color='k', normed=True)
values.plot(kind='kde', style='k--')
```

### 2.4 散点图

```python
macro = pd.read_csv('macrodata.csv')
data = macro[['cpi', 'm1', 'tbilrate', 'unemp']]
trans_data = np.log(data).diff().dropna() #取对数、差分、去空值
trans_data[-5:]
```

```python
plt.scatter(trans_data['m1'], trans_data['unemp'])
plt.title('Changes in log %s vs. log %s' % ('m1', 'unemp'))
```

可以同时绘制一组变量之间的散点图：

```python
pd.scatter_matrix(trans_data, diagonal='kde', color='k', alpha=0.3)
```
