```python
begin = Date(2018, 1, 1)
end = Date(2018, 3, 1)
```

## 1.PCR计算以及可视化

PCR: Put Call Ratio
volPCR: putVol callVol ratio
valuePCR: putValue callValue ratio
near代表当月合约
diff代表本交易日与上一个交易日ratio的差值，可以通过差值来看出市场上看涨和看跌合约之间的需求变化

```python
# 计算PCR以及PCR_Diff
PCR = opt.histPCR50ETF(begin, end)
PCR.tail(2)
```

```python
# PCR可视化
opt.plotPCR(begin, end)
```

## 2.PCVID，Greeks，IV计算以及可视化

```python
date = Date(2018, 3, 1)
```

```python
# 计算某个交易日各个期权合约Greeks和IV
Greeks_iv = opt.histDayGreeksIVOpt50ETF(date)
Greeks_iv.tail(5)
```

```python
# 画出某个交易日的四个到期月合约的隐含波动率微笑
opt.histDayPlotSmileVolatilityOpt50ETF(date)
```

```python
# 计算某个交易日的PCVID数据， near为当月， next为次月
# PCVID：diff(put_iv - call_iv), 也就是看跌IV和看涨IV差值的毎日変化
PCVID = opt.histDayPCIVD50ETF(date)
PCVID
```

```python
# PCVID可视化
opt.plotPCIVD(begin, end)
```

可以看到近期的PCVID呈上升趋势， 市场波动大啊

## 3.VIX计算以及可视化

```python
# 计算某一日的VIX数据
VIX_day = opt.getDayVIX(date)
VIX_day
```
