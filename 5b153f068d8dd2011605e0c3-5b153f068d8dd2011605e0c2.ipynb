```python
#计算因子的IC值

import pandas as pd

#1、取出2010-01-01到2017-12-31每个月末HS300股票的'ROE', 'DAVOL10'数据以及月度收益率
#思路：
# a、得到期间的月末交易日期list，注意需要去重
# b、得到对应月末的HS300成分股在该月末的因子数据和月度收益率

# a、得到期间的月末交易日期list，注意需要去重
calendar_frame = DataAPI.TradeCalGet(exchangeCD=u"XSHG,XSHE",beginDate=u"20100101",endDate=u"20171231",field=u"",pandas="1")
calendar_frame = calendar_frame[calendar_frame.isMonthEnd==1][['calendarDate', 'isMonthEnd']]
calendar_frame = calendar_frame.drop_duplicates()#因为有"XSHG,XSHE"的交易日期，所以需要去重一下，要不然每个交易日遍历两次
print calendar_frame.head().to_html()
print len(calendar_frame)
print '-'*40

# b、得到对应月末的HS300成分股在该月末的因子数据和月度收益率
factor_frame_list = []
monthly_return_list = []
for tdate in calendar_frame.calendarDate.values:
    in_to_Date = tdate.replace('-', '')#去掉日期之间的链接'-','2018-06-03'
    hs300_frame = DataAPI.IdxConsGet(secID=u"",ticker=u"000300",isNew=u"",intoDate=in_to_Date,field=u"",pandas="1")
#DataAPI.IdxConsGet获取国内外指数的成分构成情况，包括指数成分股名称、成分股代码、入选日期、剔除日期等
    tickers = list(hs300_frame.consTickerSymbol.values)
    
    factor_frame = DataAPI.MktStockFactorsOneDayGet(tradeDate=tdate,ticker=tickers,field=["secID",'tradeDate','ROE', 'DAVOL10'],pandas="1")
    factor_frame_list.append(factor_frame)#把每一天所有股票的dataframe放在一个list里面

    monthly_return_frame = DataAPI.MktEqumGet(secID=u"",ticker=tickers,monthEndDate=in_to_Date,
                                          isOpen=u"",field=["secID",'endDate','return'],pandas="1")
    monthly_return_list.append(monthly_return_frame)

tfactor_frame = pd.concat(factor_frame_list)
tfactor_frame['tradeDate'] = tfactor_frame['tradeDate'].apply(lambda x: x.replace("-",""))#去掉交易日期中的'-
print u'tfactor_frame 月度ROE, DAVOL10:\n'
print len(tfactor_frame)
print tfactor_frame.head().to_html()
print '-'*40

monthly_return_frame = pd.concat(monthly_return_list)    
monthly_return_frame['endDate'] = monthly_return_frame['endDate'].apply(lambda x: x.replace("-",""))
print u'monthly_return_frame 月度收益率data:\n'
print len(monthly_return_frame)
print monthly_return_frame.head().to_html()
```

```python
#2、进行计算IC值的因子与收益率需要错位一个月
# 思路：
# 把交易日期进行错位处理，merger具有广播功能

# a、拿到交易日历，用来做对齐,
calendar_frame1 = DataAPI.TradeCalGet(exchangeCD=u"XSHG,XSHE",beginDate=u"20091201",endDate=u"20171231",field=u"",pandas="1")
#取交易日期
calendar_frame1 = calendar_frame1[calendar_frame1.isMonthEnd==1][['calendarDate', 'isMonthEnd']]
#取出月末交易日期
calendar_frame1['calendarDate'] = calendar_frame1['calendarDate'].apply(lambda x: x.replace("-",""))
#去掉日期里面的'-'
print len(calendar_frame1)
calendar_frame1 = calendar_frame1.drop_duplicates()#由于上面选择了两个交易所的交易日期，所以去重后数量变为原来一半
calendar_frame1['prev_month_end'] = calendar_frame1['calendarDate'].shift(1)

#把交易日期往下偏移1
calendar_frame1 = calendar_frame1[['calendarDate', 'prev_month_end']]
calendar_frame1.rename(columns={"calendarDate":"month_end"}, inplace=True)
#更改列名
print u'calendar_frame1 交易日历关系\n'
print calendar_frame1.head().to_html()
print len(calendar_frame1)
print '-'*60

# b、把因子与日期merge，再与月度收益率merge
tfactor_frame1 = tfactor_frame.copy(deep=True)
tfactor_frame1.drop_duplicates()
print len(tfactor_frame1)
# 将数据对齐
tfactor_frame1 = tfactor_frame1.merge(calendar_frame1, left_on=['tradeDate'], right_on=['prev_month_end'], how='left')
#把日历dataframe与因子dataframe进行merge，日历dataframe按因子dataframe进行广播，194的长度广播到了114600
print '对齐交易日历'
print len(tfactor_frame1)
print tfactor_frame1.head().to_html()

tfactor_frame1 = tfactor_frame1.merge(monthly_return_frame, left_on=['secID', 'month_end'], right_on=['secID', 'endDate'], how='left')
print '对齐下期收益率的dataframe'
print len(tfactor_frame1)
print tfactor_frame1.head().to_html()
tfactor_frame1 = tfactor_frame1.dropna()#由于有些股票在下一期不在HS300成分股里面，所以下一期的月收益没有，所以会有空值出现，剔除空值进行IC值计算
print '剔除空值后的长度：',len(tfactor_frame1)
```

```python
#3、计算IC值
import scipy.stats as st
def calc_normal_ic(df):
    tmp_df = df.copy()
    tmp_df.dropna(inplace=True)
    return tmp_df['ROE'].corr(tmp_df['return'])

# 用scipy的包计算
def calc_normal_ic_2(df):
    tmp_df = df.copy()
    tmp_df.dropna(inplace=True)
    return st.pearsonr(tmp_df['ROE'], tmp_df['return'])[0]

# 用scipy的包计算
def calc_rank_ic(df):
    tmp_df = df.copy()
    tmp_df.dropna(inplace=True)
    return st.spearmanr(tmp_df['ROE'], tmp_df['return'])[0]

# 计算Normal IC
normal_ic_frame = tfactor_frame1.groupby(['tradeDate']).apply(calc_normal_ic)
print "Normal IC1:"
print normal_ic_frame.head()

normal_ic_frame = tfactor_frame1.groupby(['tradeDate']).apply(calc_normal_ic_2)
print "Normal IC2:"
print normal_ic_frame.head()

# 计算Rank IC
rank_ic_frame = tfactor_frame1.groupby(['tradeDate']).apply(calc_rank_ic)
print "-"*20
print "Rank IC:"
print rank_ic_frame.head()
```
