## 读研报
-----------
### *何为技术指标*
技术分析是指以市场行为为研究对象，以判断市场趋势并跟随趋势的周期性变化来进行股票及一切金融衍生物交易决策的方法的总和。目的是为了寻找买入、卖出、止损信号，并通过资金管理而达成在风险市场中长期稳定获利。
同常见的基本面指标不同，我们通过观察股票的历史价格变化和成交量成交金额的变化，处理并分析这些数据，计算出一定程度上能够反映和预测价格变化的指标，然后通过这些指标来选取股票。

### *如何评价技术指标*

一个好的因子应当具有较高的IC信息系数绝对值，较高额IR信息比，较高的组合胜率，较高的组合收益率，以及较低的换手率。
> 1. 因子IC (信息系数) ：即每个时点因子在各个股票的暴露值与各股票下个期回报的相关系数，如果一个因子的IC值高于4 %( 或低于-4%) ，则认为该因子在优选个股方面具有较好的效果，IC为正表示该因子与股票的未来收益有正相关关系，应该做多因子暴露值高的股票，反之若IC为负表示该因子与股票的未来收益负相关，则应该做空因子暴露值高的股票。
> 2. 因子IR (信息比) ：即依据该因子选择的投资组合在样本期间的平均年化收益与年化平均标准差的比值，IR 的绝对值越高，表面该因子在优选个股的收益上效果越好，若IR为正，则代表应该超配因子暴露值高的股票，反之若IR为负，则应该超配因子暴露值低的股票。
> 3. 胜率：用于衡量该技术指标因子是否在多数时间内能够获得正收益。胜率越高，说明该因子的效果越好。
> 4. 年化收益率：包括因子月平均收益和因子滚动12个月累计收益，用于衡量该技术指标因子是否具有稳定且可持续收益。此处分别考察多空组合和多头组合的收益率。
> 5. 换手率：即每个月末投资组合中被更换的股票所占原股票的比例。由于更换投资组合会产生手续费及冲击成本，因子的换手率应在保证收益率的前提下尽量低。

接下来，我们从DataAPI获取上述因子数据：

```python
'''
Preparing Factor Data from DataAPI for Calculation
'''

tech_factors = ['DAVOL10', 'DAVOL20', 'DAVOL5', 'DHILO', 'DVRAT', 'EGRO', 'EMA10', 'EMA120', 'EMA20', 'EMA5', 'EMA60', 'HSIGMA', 'MA10', 'MA120', 'MA20', 'MA5', 'MA60','MAWVAD'
,'MFI', 'PSY', 'REVS10', 'REVS20', 'REVS5', 'RSI', 'RSTR12', 'RSTR24', 'TOBT', 'VOL10', 'VOL120', 'VOL20', 'VOL240', 'VOL5', 'VOL60', 'WVAD', 'FiftyTwoWeekHigh', 'Volatility', 'Skewness', 'ILLIQUIDITY' ] # All techinal factors in DataAPI

def FactorGet(universe, trading_days):
    ret_data = {} #record the return data; key:stock_id; value: factor value in trading_days (in Pandas)
    print trading_days[0]
    for stock in universe:
        try:
            factor_data = DataAPI.MktStockFactorsDateRangeGet(secID = stock, beginDate = trading_days[0].strftime('%Y%m%d'), endDate = trading_days[-1].strftime('%Y%m%d'), field = ['tradeDate'] + tech_factors)
            factor_data['tradeDate'] = pd.to_datetime(factor_data['tradeDate'])
            print len(factor_data['tradeDate'])
        except Exception, e:
            print "Stock %s Get Factor Data Failed!" % stock
            print e
        ret_data[stock] = factor_data
    return ret_data
```

以上代码取出了universe中包含的股票在trading_days这一时间段内所有的因子数据，并返回key为stock ID， value为时间段内所有因子数据的DataFrame。

之所以这样一次性取出整个时间段的因子数据，是利用了``DataAPI.MktStockFactorsDateRangeGet``可以取单只股票连续时间段内的因子数据，减少在``handle_data``中逐天调用DataAPI带来的延迟。

但在实际回测使用中，我们需要的一天中所有股票的所有技术指标数据，因此需要利用以下函数对于数据进行一次整理：

```python
'''
Reshape the Data
Return:
key: datetime
value: pandas Index --> stock-id; column --> factors 
'''

def get_time_series_factor_data(trading_days):
    ret_dict = {}
    for cur_day in trading_days:
        day_str = cur_day.strftime('%Y-%m-%d')
        pd_dict = {}
        for stock in universe:
            stock_pd = factor_data[stock]
            temp_df = stock_pd.set_index(['tradeDate'])
            if day_str in temp_df.index: # 修正版加入！！！！！！
                pd_dict[stock] = temp_df.ix[day_str]
        ret_dict[cur_day] = pd.DataFrame(pd_dict).T
    return ret_dict
```

接下来，我们就可以用获取到的因子数据作为信号构建做多组合(long-only)，以上证50为投资域，取单个因子的风险暴露(即因子取值)的前10%，等权重买入，每20个交易日调仓。

这里值得注意的是，由于我们要遍历tech_factors列表中的38个技术指标循环进行回测，但是整个过程中回测区间、股票池等参数是不会发生变化的。在这种情况下，我们**推荐采用**quartz的``quick_backtest``的方法：

该方法以固定的``SimulationParameters``为参数，在回测区间内的行情数据只需要载入一次；同时，根据IPython Notebook的特性，建议使用如下的使用方式，即在需要多次修改策略参数的探索式研究过程中，将载入固定数据的部分(无论是行情数据还是从DataAPI中取得的数据)放在一个Cell中，而将需要修改的部分放在另一个Cell，这样固定部分的内容只需要运行一次，数据就将驻留在内存内，而不需要每次都运行，可提高效率。

```python
import quartz as qz
import pandas as pd
import numpy as np
from quartz.api import *
from datetime import datetime
from CAL.PyCAL import *
cal = Calendar('China.SSE')

start = '2011-01-05'                       # 回测起始时间
end = '2015-01-01'                         # 回测结束时间
benchmark = 'HS300'                        # 策略参考标准
universe = set_universe('SH50')
capital_base = 100000                      # 起始资金
refresh_rate = 20

data_start = '2011-01-01'
factor_days = qz.utils.tradingcalendar.get_trading_days(data_start, end)
print "Factor Get begin!"
factor_data = FactorGet(universe, factor_days)
print "Factor Get end!"

a = get_time_series_factor_data(factor_days)

sim_params = qz.sim_condition.env.SimulationParameters(start, end, benchmark, universe, capital_base)
idxmap_all, data_all = qz.sim_condition.data_generator.get_daily_data(sim_params)
```

```python
target_factor = tech_factors[0]

def win_rate(bt, capital_base):
    total_days = bt.shape[0]
    bt['prevValue'] = bt['portfolio_value'].shift(1)
    bt['prevValue'][0] = capital_base
    bt['p_return'] = bt['portfolio_value']/bt['prevValue']-1
    bt['win'] = ( bt['p_return'] > bt['benchmark_return'] )
    return ( bt['win'].values > 0 ).sum() / float(total_days)

def initialize(account):                   # 初始化虚拟账户状态
    account.pd_record = a

def handle_data(account):                  # 每个交易日的买入卖出指令
    lastTDay = cal.advanceDate(account.current_date,'-1B',BizDayConvention.Preceding)  # 获取上一个交易日的技术指标信息
    pd_data = account.pd_record[lastTDay.toDateTime()]
    order_list = list(pd_data[target_factor].dropna().order().index[-5:])
    for stock in account.valid_secpos:
        order_to(stock, 0)
    for s in order_list:
        if account.referencePrice[s] == 0:
            continue
        order(s, account.cash/account.referencePrice[s]/len(order_list))

for fac in tech_factors:
    target_factor = fac
    strategy = qz.sim_condition.strategy.TradingStrategy(initialize, handle_data)        
    bt, acct = qz.quick_backtest(sim_params, strategy, idxmap_all, data_all, refresh_rate = refresh_rate)
    perf = qz.perf_parse(bt, acct)
    out_keys = ['annualized_return', 'volatility', 'information_ratio', 'sharpe', 'turnover_rate', 'max_drawdown', 'alpha', 'beta', 'information_coefficient', 'benchmark_annualized_return']
    print "factor %s result : " % fac
    for k in out_keys:
        print '%s : %s ' % (k, perf[k])
    print 'win_rate : ', win_rate(bt, capital_base)
    print '-----------------------------------'
```

根据以上的结果，我们选取EGRO、ILLIQUIDITY和REV5作为我们在未来策略中适用的因子。

同时，需要提醒的是，我们在回测中对于挑选中的股票构建组合时，采用了简单的本金平分策略，如果希望采用更复杂的组合构建方法和更丰富的评价指标，可以把因子信号上传到[DataYes RDP](https://gw.wmcloud.com/rdp/)进行研究。

##计算新的因子
-----------
在所读研报中，经过其回测分析，向我们推荐了三大因子：
> CR指标：以上一个计算周期（如N日）的中间价比较当前周期（如日）的最高价、最低价，计算出一段时期内股价的“强弱”，从而在分析一些股价的异常波动行情时，有其独到的功能。另外，CR指标不但能够测量人气的热度、价格动量的潜能，而且能够显示出股价的压力带和支撑带，为分析预测股价未来的变化趋势，判断买卖股票的时机提供重要的参考。

    TYP = (highest + lowest + close)/3;
    CR = sum(max(highest – prev_typical, 0), N) / sum(max(prev_typical – lowest, 0), N) * 100

> PVT指标：价量趋势(PVT)指标是指很巧妙地把能量变化与价格趋势有机地联系到了一起，从而构成了量价趋势指标。参考链接：http://en.wikipedia.org/wiki/Volume-price_trend

    PVT = PVT(-N) + volume * (close - prev_close(N))/prev_close(-N)

> Aroon指标：Aroon是由Tushar Chande发明的，它通过计算自价格达到近期最高值和最低值以来所经过的期间数，帮助投资者预测证券价格从趋势到区域区域或反转的变化。参考链接：http://wiki.mbalib.com/wiki/Aroon
Aroon指标分为Aroon、AroonUp和AroonDown3个具体指标。若考察一段长度为N（一般取26个交易日）的区间，以x表示该区间最高价出现日距离当前交易日的天数，以y表示该区间最低价出现日距离当前交易日的天数。

    AroonUp = 100*(N-x)/N
    AroonDown = 100*(N-y)/N
    Aroon = AroonUp - AroonDown
