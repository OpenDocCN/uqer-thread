# 1.因子多空组合净值表现
- ### 为了之后构建策略方便，我们不妨先得到因子的多空组合净值表现，下面就是写函数来实现

```python
#加载需要使用的包
from __future__ import division
import numpy as np
import pandas as pd
import scipy.stats as st

from CAL.PyCAL import *    # CAL.PyCAL中包含font
font.set_size(22)
small_font = font.copy()
small_font.set_size(12)

import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('white')

import math
import datetime
```

```python
class factor_long_short_perf(object):
    
    def __init__(self,factors,reverse,begindate,enddate,frequency='weekly',universe_name='000906.ZICN'):
        ###----传入参数----
        self.begindate = begindate  #开始时间
        self.enddate = enddate      #结束时间
        self.frequency = frequency  #数据频率，只要日度，周度，月度三个选择
        self.universe_name = universe_name  #股票池
        self.factors = factors   #想要查看因子表现的因子名
        self.reverse = reverse #用于注明因子的效果是否是负向的
        ###----自定义参数----

        self.datelist = self.get_datelist()  #传入开始和结束日期，得到交易日列表list
        
    
    ###得到日期列表
    def get_datelist(self):   
        #从dataapi调取交易日序列
        begin_date = datetime.datetime.strptime(self.begindate,"%Y-%m-%d").strftime("%Y%m%d")
        end_date = datetime.datetime.strptime(self.enddate,"%Y-%m-%d").strftime("%Y%m%d")
        date_data=DataAPI.TradeCalGet(exchangeCD=u"XSHG",beginDate=begin_date,endDate=end_date,field=['calendarDate','isWeekEnd','isMonthEnd'])
        if self.frequency == 'daily':   #按照规定的频率处理得到的交易日序列
            #self.M = 250
            date= map(lambda x: x[0:4]+x[5:7]+x[8:10], date_data['calendarDate'].values.tolist())
        if self.frequency == 'weekly':
            #self.M = 52
            date_data = date_data[date_data['isWeekEnd'] == 1]
            date= map(lambda x: x[0:4]+x[5:7]+x[8:10], date_data['calendarDate'].values.tolist())
        if self.frequency == 'monthly':
            #self.M = 12
            date_data = date_data[date_data['isMonthEnd'] == 1]
            date= map(lambda x: x[0:4]+x[5:7]+x[8:10], date_data['calendarDate'].values.tolist())
        return date
    
    #快速回测
    def quick_test(self,factor_name):
        #快速回测
        # 可编辑部分与 strategy 模式一样，其余部分按本例代码编写即可

        # -----------回测参数部分开始，可编辑------------



        start = self.begindate                       # 回测起始时间
        end = self.enddate     # 回测结束时间
        universe = DynamicUniverse(self.universe_name)     # 股票池
        benchmark = 'HS300'                       # 策略参考标准
        capital_base = 30000000                     # 起始资金
        max_history_window=50


        # 日期处理相关
        cal = Calendar('China.SSE')
        period = Period('-1B')


        # ---------------回测参数部分结束----------------

        # 把回测参数封装到 SimulationParameters 中，供 quick_backtest 使用
        sim_params = quartz.SimulationParameters(start, end, benchmark, universe, capital_base, max_history_window=max_history_window)
        # 获取回测行情数据
        data = quartz.get_backtest_data(sim_params)
        # 运行结果
        results_quantile = {}
        # 调整参数快速回测
        for x in range(2):

            # -----------策略逻辑部分开始，可编辑------------

            # refresh_rate 的值和回测行情数据无关，所以也可以在参数调优中修改
            # 类似的还有 commission 和 slippage
            refresh_rate = 1
            # commission = ...
            # slippage = ...

            def initialize(account):
                pass
            def handle_data(account):
                today = account.current_date
                today = Date.fromDateTime(account.current_date)  # 向前移动一个工作日
                yesterday = cal.advanceDate(today, period)
                yesterday = yesterday.toDateTime().strftime('%Y%m%d')
                if yesterday in self.datelist:
            
            
                    factor = DataAPI.MktStockFactorsOneDayGet(tradeDate=yesterday,secID=account.universe,field='secID,'+factor_name).dropna()
                    factor.set_index('secID',inplace=True)
                    
                    if x == 0:
                        quantile20=np.percentile(factor[factor_name],20)   #五分位数组合
                        factor=factor[factor[factor_name]<=quantile20]
                    if x == 1:
                        quantile80=np.percentile(factor[factor_name],80)
                        factor=factor[factor[factor_name]>=quantile80]
        
                    # 先卖出
                    sell_list = account.security_position
                    for stk in sell_list:
                        order_to(stk, 0)

                    # 再买入
                    buy_list = factor.index
                    total_money = account.reference_portfolio_value
                    prices = account.reference_price 
                    for stk in buy_list:
                        if np.isnan(prices[stk]) or prices[stk] == 0:  # 停牌或是还没有上市等原因不能交易
                            continue
                        order(stk, int(total_money*1.0 /len(buy_list)/ prices[stk] /100)*100)
                else:
                    return

            # ---------------策略逻辑部分结束----------------

            # 把回测逻辑封装到 TradingStrategy 中，供 quick_backtest 使用
            strategy = quartz.TradingStrategy(initialize, handle_data)
            # 回测部分
            bt, perf = quartz.quick_backtest(sim_params, strategy, data=data, refresh_rate=refresh_rate)
            # commission = commission,
            # slippage = slippage

            # 保存运行结果
            results_quantile[x] = perf
        return results_quantile
    
    #对每个因子作以上过程，即看多空组合表现
    def factors_perf(self):
        results = pd.DataFrame()
        for i in range(len(self.factors)):
            results_quantile = self.quick_test(self.factors[i])
            result = pd.concat([results_quantile[0]['cumulative_returns'],results_quantile[1]['cumulative_returns']],axis=1)
            result.columns = [self.factors[i],'80']
            if self.reverse[i] == True:
                result[self.factors[i]] = result[self.factors[i]]-result['80']
            else:
                result[self.factors[i]] = result['80']-result[self.factors[i]]
            del result['80']
            results = pd.concat([results,result],axis=1)
        results.index = map(lambda x: Date.fromDateTime(x).toDateTime().strftime('%Y%m%d'), results.index)
        return results+1
    
    #把净值曲线做个图，嘿嘿
    def perf_plot(self,data):
        fig = plt.figure(figsize = (16,6))
        ax = fig.add_subplot(111)
        data.plot(ax=ax,title=u'因子多空组合净值表现')
        #ax.legend(map(lambda x: x.decode('utf-8'),data.columns.values),prop=small_font,loc='best')
        ax.set_xlabel(u'交易日期', fontproperties=small_font)
        plt.xticks(fontproperties = small_font, rotation = 70)
        ax.title.set_font_properties(font)
        ax.set_axis_bgcolor('white')
        ax.grid(color='gray', alpha=1.0, axis='y')
```

- ### 输入：
> #### 因子名（列表）
> #### 因子作用方向(False表示正向的，即因子值越大，股票收益越高)
> #### 研究时间范围：开始，结束
> #### 频率：默认为周度
> #### 股票池名称：默认为中证800成分股

- ### 输出：
> #### 因子多空组合净值表现（DataFrame）
> #### 作图

```python
#参数设置
factor_name = ['LCAP','PB','NetAssetGrowRate','ROE','VOL60']
test = factor_long_short_perf(factor_name,[True,True,False,False,True],'2014-08-01','2016-08-01')
```

# 2.策略详情
- ### 下面详细介绍下，策略的详细思路：

> ### 在每个交易日计算过去N个交易日（包括当前交易日）因子多空组合净值的均值C，同时计算过去N个交易日因子多空组合收益的均值R和标准差$\sigma$,从而得到多空净值上界$Upper = C*(R+2*\sigma)$和多空净值下界$Lower = C * (R-2*\sigma)$。
> ### 首先，赋予因子初始权重。
> #### 当因子多空净值向上穿越了上界时，我们便提升该因子的权重，权重变为初始权重的2倍；当因子多空净值向下穿越了下界时，我们便降低该因子的权重，权重变为初始权重的一半；而当该因子的多空净值位于上下界之间时，我们则对该因子未来表现持中性态度，权重不变。
> ### 最后对因子权重作归一化处理，根据每周最后一个交易日的因子权重对[分位数标准化](https://en.wikipedia.org/wiki/Quantile_normalization)之后的因子数据进行合成，得到新的合成因子

# 三.总结
- ### 可以看到，在未对因子做中性化处理，策略构建时也没有考虑行业分散，止损等优化方法,甚至参数也没有去拟合(网太烂。。。)的情况下，因子轮动策略的表现还是稳扎稳打的
- ### 本帖重在思路，许多细节并没有处理的尽善尽美，比如因子的选取上，也没有仔细斟酌。
- ### 要打完全套武功，还需要对因子进行充分挖掘后，选取几个大类因子，对每个类里的因子进行合成；构建策略时，还应当考虑因子的行业中性化和市值中性化处理，并且投资组合应当考虑行业分散化或者更全面的考察风险集中在哪些因子上；如何综合使用诸多方法，是策略构建的重点，还需继续研究。。。
