一：动态演化：趋同度择时的新视角
选取上证综指2010-01-04至2016-12-12每日指数收盘价与趋同度指数，将趋同度周涨跌幅与指数周涨跌幅画成散点图，其中，横轴为趋同度当周涨跌幅，纵轴为指数当周涨跌幅（为避免路径依赖，此处周涨跌幅采用每五日轮动计算周涨跌幅，其算法是对每一天循环　计算今天的和五天前的数据的收益率　这样可以避免由于路径产生的随机性）

```python
import numpy as np
import pandas as pd
import talib as ta
import datetime
from dateutil.parser import parse#修改时间格式的作用
import matplotlib.pyplot as plt
from CAL.PyCAL import *    # CAL.PyCAL中包含font
font.set_size(15)
ticker='000001'
begin='20100103'
end='20161213'
M=30
filename='convergence_30_shangzheng.csv'
frequency=6#这个参数是　由于周涨幅　滚动的５日收益率
```

```python
#获取趋同度数据　
convergence_data = pd.read_csv(filename,header=None)
convergence_data = convergence_data.rename(columns={0:'tradeDate',1:'convergence_30'})
convergence_data['tradeDate'] = convergence_data['tradeDate'].map(lambda x:parse(x))#日期转换格式 注意
```

```python
#获取指数数据
df_index = DataAPI.MktIdxdGet(ticker=ticker,beginDate=begin,endDate=end,field=u"tradeDate,preCloseIndex,closeIndex",pandas="1")
#df_index['index_ret'] = df_index['closeIndex'] / df_index['preCloseIndex'] - 1
df_index['tradeDate'] = df_index['tradeDate'].map(lambda x:parse(x))#拿指数 计算30天的收益
```

```python
def convergence_index_plot(convergence_data,df_index,start_date,end_date):
    #画趋势图与指数行情图
    start_date = start_date if isinstance(start_date,datetime.datetime) else parse(start_date)#判断一个对象是否是已知类型 
    #这个写法就是 判断是datetime.datetime 类型  如果不是就改过去
    end_date = end_date if isinstance(end_date,datetime.datetime) else parse(end_date)
    convergence_data_sample = convergence_data[(convergence_data['tradeDate']>=start_date) & (convergence_data['tradeDate']<=end_date)]
    df_index_sample = df_index[(df_index['tradeDate']>=start_date) & (df_index['tradeDate']<=end_date)]
    fig = plt.figure(figsize=(18,7))
    ax = fig.add_subplot(111)#加子图 下面的指令应该是画在一起
    ax2=ax.twinx()
    ax.plot(df_index_sample['tradeDate'].values,df_index_sample['closeIndex'].values,'darkgoldenrod',linewidth=1.5)
    ax2.plot(convergence_data_sample['tradeDate'].values,convergence_data_sample['convergence_30'].values,'crimson',linewidth=1.5)
    #legend是加线的名字
    ax.legend([ticker],loc='upper left',bbox_to_anchor=(0.01,0.99),prop=font,frameon= False,handlelength = 3,handletextpad=0.1,borderpad = 0.1)
    ax2.legend([u"趋同度"],loc='upper left',bbox_to_anchor=(0.01,0.92),prop=font,frameon= False,handlelength = 3,handletextpad=0.1,borderpad = 0.1)
    ax.spines['top'].set_color('none')
    ax2.spines['top'].set_color('none')
    ax.xaxis.set_ticks_position('bottom')
    # plt.ylim(df_index_sample['closeIndex'].min(),df_index_sample['closeIndex'].max())
    #ax.set_title(u"图1 中证全指趋同度", fontproperties=font, fontsize=18)
```

下面开始计算五日滚动的收益率

```python
indexret_5 = pd.rolling_apply(df_index['closeIndex'],frequency,lambda x:x[-1]/x[0] - 1)
indexret_5 = indexret_5.iloc[1:].reset_index(drop=True)#指数的周收益率　（每天都有数据）
convergenceret_5=pd.rolling_apply(convergence_data['convergence_30'],frequency,lambda x:x[-1]/x[0] - 1)
convergenceret_5 = convergenceret_5.iloc[1:].reset_index(drop=True)#趋同度的收益率
convergence_indexret = pd.concat([convergenceret_5,indexret_5],axis=1)
convergence_indexret = convergence_indexret.dropna()
convergence_indexret.columns =  ['convergence_30_5','index5ret'] # 整合指数收益率与趋同度
```

```python
convergence_indexret
```

```python
convergence_indexret.index=df_index['tradeDate'][convergence_indexret.index[0]:-1]
```

```python
convergence_indexret#每一天的滚动五日涨跌幅
```

```python
#2007年到 2017年的每一个周五交易日的日期 
data=DataAPI.TradeCalGet(exchangeCD=u"XSHG",beginDate=u"20091001",endDate='20170301',field=['calendarDate','isWeekEnd'],pandas="1")
#data = data[data['isMonthEnd'] == 1]
data = data[data['isWeekEnd'] == 1]
date_list1 = map(lambda x: x[0:4]+x[5:7]+x[8:10], data['calendarDate'].values.tolist())
print date_list1
```

实际上这里面存在一个问题 就是你不可以认为每周期间有五个数据 特殊情况该如何处理  比如周一周二放假这周就只有三个数据  （一种方法就是不管特殊情况　就是找之前的五天平均
还有就是按照实际的周进行平均）　（这里为了简单选取第一种方式）

```python
#采用五天的方法计算　　按道理来讲指数的周收益不应该采用滚动平均的方式　可以直接取数据　　但是研报里的计算是滚动平均的　这里也采用滚动的
comb=convergence_indexret
s=comb.columns
week_date=[]
week_re=[]
for i in range(0,len(comb)-5):
    if comb[i:i+1].index in date_list1:#date_list1是实际的周末数据
        sum_temp=0
        for j in range(0,5):
            sum_temp+=comb[[s[0],s[1]]][i+1+j:i+2+j].values
        week_re.append(sum_temp[0]/5.0)
        week_date.append(comb[i+5:i+6].index.values[0])
```

```python
final_data=pd.DataFrame(np.nan,index=week_date,columns=[s[0],s[1]])
for i in range(0,len(final_data)):
    final_data[i:i+1]=week_re[i:i+1]
final_data#每个周五的特征数据　
```

接下来划分象限　每个象限代表着不同的情况

```python
label=[]#这个是区分象限的   趋同度周涨幅为x轴 指数周涨幅为y轴   
s=final_data.columns
for i in range(0,len(final_data)):# 16 (4*4)
    if final_data[s[0]][i:i+1].values>0:#x axis
        if final_data[s[1]][i:i+1].values>0:
            label.append(1)
        else:
            label.append(4)
    else:
        if final_data[s[1]][i:i+1].values>0:
            label.append(2)
        else:
            label.append(3)
final_data['label']=label
```

关于四个象限的分析可以参考卢威老师的https://uqer.datayes.com/community/share/57fee2e1228e5b3668facc70（尽管卢老师的帖子里区分的是趋同度的数值进行分类但是仍有参考价值）
下面是研报中的解释：
	第I象限：指数上涨，趋同度指数上涨。大盘走高且个股日渐趋同，意味买盘信心增加。故指数大概率将继续上涨，为看多信号。 
    第II象限：指数上涨，趋同度指数下跌。虽然大盘涨势未止，但个股趋于分化，多头信心动摇。故上涨难以持续，为看空信号。 
    第III象限:指数下跌，趋同度指数下跌。与第II象限类似，虽然大盘仍处于跌势，但个股趋于分化，空头信心动摇。故下跌难以持续，为看多信号。 
    第IV象限：指数下跌，趋同度指数上涨。与第I象限类似，大盘下跌且个股日渐趋同，意味着卖盘信心增加。故指数大概率将继续下跌，为看空信号
（但是我们作出的结果好象不是这么回事　感觉结果与Ｍ的值有很大关系　Ｍ越小趋同度的变化越快　极值也越大　可以适当减少Ｍ的大小）
    
下面介绍：象限间的状态转移矩阵。
	首先他是一个M行N列的矩阵 第M行第N列的数值，代表第t周处于第M象限，而第t+1周处于第N象限的概率（M=1,2,3,4; N=1,2,3,4）。 第M行第5列是 T时在 某个象限（对应四行当中的某一行M） T+1时的大盘上涨 第六列是大盘下跌　（当然也可以做月度的转移矩阵或者是每天的转移矩阵　不过每天的转移矩阵可以预见它的对角元素的值会很大
因为每一天的状态来不及切换很大概率效仿之前的状态）

```python
ma_change=np.zeros((4,6))
s=final_data.columns
for i in range(1,len(final_data)-1):
    s1=final_data[s[2]][i-1:i].values[0]#T时的 label
    s2=final_data[s[2]][i:i+1].values[0]#T+1 时的 label
    #print s1
    ma_change[s1-1][s2-1]+=1#相应的位置计数
    if final_data[s[1]][i:i+1].values[0]>0:
        ma_change[s1-1][4]+=1#next week up 意思是T时在 某个象限（对应四行当中的某一行） T+1时的大盘上涨 
    else:
        ma_change[s1-1][5]+=1# next week down
print ma_change#计数后的矩阵 

for i in range(0,4):
    temp1=ma_change[i][0]+ma_change[i][1]+ma_change[i][2]+ma_change[i][3]
    ma_change[i][0:4]=ma_change[i][0:4]/temp1
    temp=ma_change[i][4]+ma_change[i][5]
    ma_change[i][4]=ma_change[i][4]/temp
    ma_change[i][5]=ma_change[i][5]/temp
print ma_change
```

上面的矩阵 index是 1234 象限 columns是 1234象限和次周上涨概率 次周下降概率　（不过此处的结果和研报中的结果不同，按此试验的结果应该是第一第二象限做多
如果大家有兴趣的话请可以对每周实际天数进行平均）
（其实根据这个状态转移矩阵就大体上预测了下面的结果图）
下面会分析根据趋同度进行的指数择时验证： 
研报中给出的择时策略： 做多信号： 本周状态为第一象限或第三象限，次周做多，持仓时间为一周。本周状态为二或四则下周不进行开仓操作且平掉手里的全部仓位　

回测架构：依据指数的收益率来计算　（关于为什么不可以按照指数基金的买卖利用平台回测：有待讨论）
比如这周的ｌａｂｅｌ是１那么就记录下一周的实际指数收益率　如果label不是１就继续向下循环　　这样每一个记录下的收益率都是我们根据这周的数据对下周进行预测而得到的实际收益率　（我们可以对label 进行分析　可以让label 为１或２　。。。）

```python
final_data#根据这个数据的ｌａｂｅｌ　来进行择时
```

```python
temp_1=DataAPI.MktIdxwGet(beginDate=begin,endDate=end,indexID=u"000001.ZICN",ticker="000001",field=u"endDate,chgPct",pandas="1")
temp_1['endDate']=pd.to_datetime(temp_1['endDate'], format='%Y-%m-%d')#这里修改日期格式　方便匹配
temp_1.index=temp_1['endDate']
```

```python
temp_1#实际指数的周收益率
```

```python
re_ser_1=[]
re_ser_2=[]
re_ser_3=[]
re_ser_4=[]
for i in range(0,len(final_data)-1):
    if final_data.index[i] in temp_1.index:#判断当周的日期
        if final_data[i:i+1]['label'].values[0]==1 :
            if final_data.index[i+1] in temp_1.index:#判断次周的日期
                re_ser_1.append(temp_1.loc[final_data.index[i+1]]['chgPct'])#记录次周的实际涨幅
        if final_data[i:i+1]['label'].values[0]==2 :
            if final_data.index[i+1] in temp_1.index:
                re_ser_2.append(temp_1.loc[final_data.index[i+1]]['chgPct'])
        if final_data[i:i+1]['label'].values[0]==3 :
            if final_data.index[i+1] in temp_1.index:
                re_ser_3.append(temp_1.loc[final_data.index[i+1]]['chgPct'])
        if final_data[i:i+1]['label'].values[0]==4 :
            if final_data.index[i+1] in temp_1.index:
                re_ser_4.append(temp_1.loc[final_data.index[i+1]]['chgPct'])
re_1=[re_ser_1[0]+1]
re_2=[re_ser_2[0]+1]
re_3=[re_ser_3[0]+1]
re_4=[re_ser_4[0]+1]
for i in range(1,len(re_ser_1)):
    re_1.append(re_1[-1]*(re_ser_1[i]+1))#开始都是一个单位　进行累乘
for i in range(1,len(re_ser_2)):
    re_2.append(re_2[-1]*(re_ser_2[i]+1))
for i in range(1,len(re_ser_3)):
    re_3.append(re_3[-1]*(re_ser_3[i]+1))
for i in range(1,len(re_ser_4)):
    re_4.append(re_4[-1]*(re_ser_4[i]+1))
plt.figure(figsize=(14,6))
plt.plot(range(0,len(re_1)),re_1,label="label_1")
plt.plot(range(0,len(re_2)),re_2,label="label_2")
plt.plot(range(0,len(re_3)),re_3,label="label_3")
plt.plot(range(0,len(re_4)),re_4,label="label_4")
plt.legend(loc='upper left')
```

可以看出　对于label_２是一个很好的择时因子　label_3 是一个很好的反向择时因子

```python
re_ser_comb=[]
for i in range(0,len(final_data)-1):
    if final_data.index[i] in temp_1.index:
        if final_data[i:i+1]['label'].values[0]==2 or final_data[i:i+1]['label'].values[0]==1 or final_data[i:i+1]['label'].values[0]==4:#对１２3做多
            if final_data.index[i+1] in temp_1.index:
                re_ser_comb.append(temp_1.loc[final_data.index[i+1]]['chgPct'])
        if final_data[i:i+1]['label'].values[0]==3 :#如果可以做空的话　就对３进行做空操作
            if final_data.index[i+1] in temp_1.index:
                re_ser_comb.append(-temp_1.loc[final_data.index[i+1]]['chgPct'])
re_comb=[re_ser_comb[0]+1]
for i in range(1,len(re_ser_comb)):
    re_comb.append(re_comb[-1]*(re_ser_comb[i]+1))
plt.figure(figsize=(14,6))
plt.plot(range(0,len(re_comb)),re_comb,label="label_comb")
index_com=DataAPI.MktIdxwGet(beginDate=begin,endDate=end,ticker=u"000001",field=u"preClosePrice,endDate",pandas="1")
index_com=index_com.sort(columns='endDate')
index_com.index=range(len(index_com))
(index_com['preClosePrice']/3276.0).plot(label='index')
plt.legend(loc='upper left')
```
