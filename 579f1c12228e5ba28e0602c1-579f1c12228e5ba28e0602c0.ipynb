## **1,综述**
近日研究了东北证券的研报“偏度调整的收益率 MACD 择时信号”觉得十分巧妙，本文希望根据其思想做一个相关的研究。<br>

我们在分析过程中仍然采用回复 S 型函数对不同周期现货收益率的 EMA 差值（EMA2）进行了变换，以确定触发择时信号 EMA2 是平稳增加而非由于单一冲击因素带来异常变动，起到了信号除噪的作用。此外，我们对数据的偏度进行了观察，发现即使在长期看来收益率以及其对应的移动平均数据的偏度为负，但在中短期却时常出现正偏度，并且偏度具有一定的连续型，及在数值上不会在短时间出现较大的变化。这对我们在选择信号点时带来了一定的指导性，在基于收益率构造择时信号时，应给予数据偏度更多的关注，以确定当前市场环境适合的信号参数。<br>

这种方法的核心思想是追踪平稳趋势，并尽量在震荡行情保持小幅收益。


## **2,模型介绍**
定义$X_n$为标的资产价格，定义$U_n$为资产第 n 期的收益率，即$X_n=\frac{U_{n+1}}{U_n}$<br>

于是，我们采用如下方法计算EMA：<br>
<br>
$$EMA_α[X]_n=(1-α)Σ^∞_{j=0}α^jU_{n-j}$$

接着我们继续构建基于EMA指标的EMA2指标：<br>
<br>
$$EMA2(α,β,n)=EMA_α[X]_n/w_α-EMA_β[X]_n/w_β$$
$$=\frac{(1-α)Σ^∞_{j=0}α^jU_{n-j}}{w_α}-\frac{(1-β)Σ^∞_{j=0}β^jU_{n-j}}{w_β}$$
其中：
$$α=1-\frac{1}{N_α},β=1-\frac{1}{N_β}\ \ \ N_α<N_β$$
为长期衰老因子。<br>
$$w_α=1-α^n,w_β=1-β^n$$
为权重因子。<br>

```python
import matplotlib.pyplot as plt
from matplotlib import rc
rc('mathtext', default='regular')
import seaborn as sns
sns.set_style('white')
from matplotlib import dates
import datetime
import numpy as np
from numpy import *
import pandas as pd
from scipy.stats import *
from pandas import Series
import time
import scipy.stats as st
from CAL.PyCAL import *    # CAL.PyCAL中包含font
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
```

```python
def EMA(U, alpha,beta):
    epsilon = 0
    result = [None]*len(U)
    for i in range(len(result)):
        w1=w2=1.0
        va=vb=0
        d1=d2=0
        for u in U[i::-1]:
            va+=u*w1
            d1+=w1
            w1*=alpha
            vb+=u*w2
            d2+=w2
            w2*=beta
            if w2<epsilon or w1<epsilon: 
                break
        
        result[i] = (va/d1-vb/d2)*1000

    return result
```

这里我们定义回复S型(Reverting Sigmoid)函数Ψ：<br>
$$Ψ(z)=c_λ*z*exp(\frac{-λ^2z^2}{2})$$

本期选用的函数参数为λ=0.5,$c_λ=1.5^\frac{3}{4}$。这个变换对于突发性增长的EMA2 进行了惩罚，当原始动量数值超过函数最高点对应的数值后，Ψ变成减函数，因此当市场受到冲击而发生较大变化时，Ψ(z) 将越过开仓阈值，以避免直接采用EMA2 进行择时可能的误导性信号.

<br>
从下面的图中可以看出在正半轴，Ψ(z) 在 0~2 之间为增函数，2 以上为减函数，且以 2 为轴左右不对称，呈现类似于负偏的形态。当 Z=EMA2 从小变大使得Ψ(z) 自下而上穿越阈值，并落入安全范围，视为做多信号，继而维持在阈值以上视为平稳增加，经历趋势延续并重新自上而下越过阈值视为信号结束；若 Z=EMA2 对应的Ψ(z) 从阈值以下受到冲击而越过安全范围到达右侧，则无法触发开仓信号。在负半轴同理。


```python
def phi(z):
    U=(9/8)**(3/4)*z*2.71**(-z*z/8)
    return U
```

```python
zArray = np.linspace(-10, 10, 100)
yArray = np.linspace(-10, 10, 100)
for i in range(len(yArray)):
    yArray[i]=phi(zArray[i])
plt.plot(zArray,yArray)
plt.xlim(-10,10)
plt.title(u"取值范围[-10,10]，λ=0.5 的Ψ(z)分布图", fontproperties=font, fontsize=16)
plt.show()
```

选择 EMA2 的主要原因是，直接基于价格的均线系统则相对滞后，而收益率变化在趋势行情中有较好的预警性。下图展示了上涨情况下的EMA和价格变动关系：在趋势形成初期，收益率的EMA由负转正，在行情中段，收益率的EMA平稳增加，且在市场走势最为稳定时达到较大值。当趋势行情接近尾声，收益率的EMA投资策略报告减少，先于行情提前反转并向下穿越 0 轴，因此在趋势行情尾段鲜再出现做多信号，并且常常在其由正转负时市场发生反转。

下面我们就进行偏度分析，Samuelson 在 1970 年就已经发现，在最优风险决策函数中加入三阶或更高阶矩之后，相比于均值-方差效用函数解决方案得以进一步完善。在数据回测和分析过程中，我们发现即使收益率偏度在长期来看的确为负，但在中短期内却不尽然，甚至有可能达到较大的正值。

```python
def skewness(time):
    Price=DataAPI.MktIdxdGet(tradeDate=u"",indexID=u"",ticker=u"000300",beginDate=time-datetime.timedelta(days=60),endDate=time,field=u"",pandas="1").closeIndex
    skew=pd.Series(len(Price)-1)
    Rate=pd.Series(len(Price)-1)
    for i in range(len(Price)-1):
        Rate[i]=np.log(Price[i+1]/Price[i])
    y=EMA(Rate,0.9,0.966)
    for i in range(len(y)):
        skew[i]=((y[i]-np.mean(y))/np.std(y))**3
    SK=np.mean(skew)
    return SK
```

```python
DATA=pd.DataFrame(index=range(48),columns=[0,1])
k=0
i=0
t1=datetime.datetime(2004,1,1)
while(k<48):
    t1=t1+datetime.timedelta(days=i)
    i=i+1
    if (t1.month==1 or t1.month==4 or t1.month==7 or t1.month==10) and t1.day==1:
        for j in range(100):
            t2=t1+datetime.timedelta(days=j)
            if (t2.month==3 and t2.day==31) or (t2.month==6 and t2.day==30) or (t2.month==9 and t2.day==30) or (t2.month==12 and t2.day==31):
                P=DataAPI.MktIdxdGet(tradeDate=u"",indexID=u"",ticker=u"000300",beginDate=t1,endDate=t2,field=u"",pandas="1").closeIndex
                skew=pd.Series(len(P)-1)
                Rate=pd.Series(len(P)-1)
                for i in range(len(P)-1):
                    Rate[i]=np.log(P[i+1]/P[i])
                y=EMA(Rate,0.9,0.966)
                for i in range(len(y)):
                    skew[i]=((y[i]-np.mean(y))/np.std(y))**3
                SK=np.mean(skew)
                DATA[0][k]=SK
                DATA[1][k]=(P[len(P)-1]-P[0])/P[0]
                k=k+1
                i=0
                break
    t1=t2
DATA
```

此外，偏度变化的范围也较大，就 2004 年 2016 年 2 月的 HS300季度偏度的计算结果来看，偏度范围为[-2.09,1.86]。

```python
NMB=pd.DataFrame(index=range(5),columns=range(5))
for i in range(5):
    for j in range(5):
        NMB[i][j]=0
for i in range(48):
    NMB[k[0][i]][k[1][i]]=NMB[k[0][i]][k[1][i]]+1.0
NMB
```

下图展示了偏度大小和季节涨跌幅的相关关系，横轴数值大小表示偏度大小的分组，纵轴数值大小表示季度涨跌幅大小的分组，颜色越深表示重合度越高，较大的正偏度不一定伴随有较大的行情，但较大的趋势行情往往伴随有绝对值较大的负偏度，并且伴随着偏度的向某一个方向绝对数值增大，其与市场走势的反向相关性越强。直观上理解，在偏度为较大的正值时，均值右侧的小部分较大的绝对正收益与均值左侧的大部分较小的收益（或正或负）相对，偏度为绝对值较大的负值时，均值左侧的小部分较大负收益与均值右侧的大部分较小收益（或正或负）相对。因此，在偏度向某个方向继续深化变或者数值很大时，对于当前阶段的择时选择应予以特别关注，即使是在震荡行情也可能会有部分日期出现较大的涨跌。

## **3，股指检验**
下面我们将这一模型在实际的市场中进行检验，我们分别使用了HS300和ZZ500两各直属进行检验。本文模型选择了$N_α=10，N_β=30$安全范围最高点对应Ψ(z)最大值Ψ(z)= 1 .2 。安全范围根据前 60 天的 EMA2 的偏度进行调整。当偏度数值较大时，调高做多开仓阈值，调低做空开仓阈值，当偏度负向较大时，负向调大做空开仓阈值，调低做多开仓阈值，当偏度处于 0 点附近时，多空开仓阈值对称选择。

```python
def Thresh(SK):
    t1=-2
    t2=2
    if SK<-1.2:
        t1=-2.5
        t2=1.5
    elif SK>1.2:
        t1=-1.5
        t2=2.5
    return t1,t2
```

```python
P=DataAPI.MktIdxdGet(tradeDate=u"",indexID=u"",ticker=u"000300",beginDate=u"20040101",endDate=u"20160101",field=u"",pandas="1")
Q=pd.Series(len(P)-1)
ema=pd.Series(len(P)-1)
for i in range(len(P)-1):
    Q[i]=np.log(P.closeIndex[i+1]/P.closeIndex[i])
ema=EMA(Q,0.9,0.966)
plt.title(u"HS300 EMA波动", fontproperties=font, fontsize=16)
plt.plot(ema)
```

下图显示了根据模型找到的买入和卖出信号，红色表示买入信号出现，绿色表示卖出信号出现。（将注释中的‘#’去掉运行可以输出全部买入卖出信号开始和结束的时间。）可以看到信号的表现还是很好的，能够较准确地识别出上涨和下跌的发生。然而缺点也是十分明显的，在股票快速上涨的时候，模型会报出大量的做空信号，这种信号明显和真实情况不符（回测中主要表现在2008和2015两次大涨的时候），这是一个值得改进的地方。

```python
fig = plt.figure(figsize=(20,6))
plt.plot(P.closeIndex)
plt.xlim(0,3000)
for i in range(len(ema)-1):
    time_struct=time.strptime(P.tradeDate[i],"%Y-%m-%d")
    date_time=datetime.datetime(time_struct.tm_year,time_struct.tm_mon,time_struct.tm_mday)
    SK=skewness(date_time)
    t1,t2=Thresh(SK)
    if ema[i]<t1 and ema[i+1]>t1 and phi(ema[i+1])>-1.2:
#        print ['buy','start',date_time]
        plt.scatter(x=array(i),y=array(P.closeIndex[i]),c='r')
    elif ema[i]>t2 and ema[i+1]<t2 and phi(ema[i+1])<1.2:
#        print ['sell','start',date_time]
        plt.scatter(x=array(i),y=array(P.closeIndex[i]),c='green')
#    elif ema[i+1]<t1 and ema[i]>t1 and ema[i]<t2:
#        print ['buy','end',date_time]
#    elif ema[i+1]>t2 and ema[i]<t2 and ema[i]>t1:
#        print ['sell','end',date_time]
plt.show()
```
