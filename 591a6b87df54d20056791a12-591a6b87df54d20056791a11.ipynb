# 三因子数据收集

```python
import numpy as np
import pandas as pd 
from CAL.PyCAL import *

Year = 2014
data = DataAPI.TradeCalGet(exchangeCD=u"XSHG",beginDate=str(Year*10000+1201),endDate=str((Year+2)*10000+1231),field=['calendarDate','isMonthEnd'])
data = data[data['isMonthEnd'] == 1]
date= map(lambda x: x[0:4]+x[5:7]+x[8:10], data['calendarDate'].values.tolist())
monthdate = list(data.calendarDate)

def Areturn(i = 0):
    breakpoint = date[i]                      
    #取沪深300
    universe = set_universe('HS300',date=breakpoint)
   
    ##SMB股票选取市场规模因子
    ME = DataAPI.MktEqudGet(tradeDate=breakpoint,secID=universe,field=u"ticker,marketValue").dropna()   
    ME30 = np.percentile(ME['marketValue'],30) 
    ME70 = np.percentile(ME['marketValue'],70)
    SM = ME[ME['marketValue']<=ME30]['ticker'].tolist()                               
    BM = ME[ME['marketValue']>ME70]['ticker'].tolist()
    ##HML股票选取市账率因子
    BP = DataAPI.MktStockFactorsOneDayGet(tradeDate=breakpoint,secID=universe,field=u"ticker,PB").dropna() 
    BP = BP[BP>0].dropna()                                                
    BP[['PB']]=1/BP[['PB']]                                                
    BP30 = np.percentile(BP['PB'],30)
    BP70 = np.percentile(BP['PB'],70)
    LP = BP[BP['PB']<=BP30]['ticker'].tolist()                                      
    HP = BP[BP['PB']>BP70]['ticker'].tolist()
    
    tickerlist = [SM,BM,LP,HP]
    tickername = ['SM','BM','LP','HP']
    #取出选择股票计算收益率
    cfrm = pd.DataFrame()
    for n,j in zip(tickername,tickerlist):
        a = DataAPI.MktEqudGet(tradeDate=date[i],secID=u"",ticker= j ,beginDate=u"",endDate=u"",isOpen="",field=u"ticker,	tradeDate,accumAdjFactor,closePrice",pandas="1")
        a['price'] = a['closePrice'] * a['accumAdjFactor']
        b = DataAPI.MktEqudGet(tradeDate=date[i+1],secID=u"",ticker= j ,beginDate=u"",endDate=u"",isOpen="",field=u"ticker,	tradeDate,accumAdjFactor,closePrice",pandas="1")
        b['price'] = b['closePrice'] * b['accumAdjFactor']
        c = ((b['price'] - a['price'])/a['price']).mean()
        cfrm[n] = [c]
    return cfrm

Returnfrm = pd.DataFrame()
for t in range(len(monthdate)-1):
    x = Areturn(t)
    Returnfrm = pd.concat([Returnfrm,x])
Returnfrm.index = monthdate[1:]

SMB = Returnfrm['SM'] - Returnfrm['BM']
HML = Returnfrm['LP'] - Returnfrm['HP']


#大盘指数
Rc = DataAPI.MktIdxdGet(tradeDate=u"",indexID=u"",ticker=u"000001",beginDate=date[0],endDate=date[-1],field=u"tradeDate,closeIndex",pandas="1")
Rc.index = list(Rc.tradeDate)

Rc['Index'] = Rc.ix[list(monthdate)]['closeIndex']
Rc = Rc.dropna()
Rturn = Rc['Index'].pct_change().dropna()-0.0025


P_r1 = DataAPI.MktEqudGet(tradeDate=u"",secID=u"",ticker=u'600050',beginDate=date[0],endDate=date[-1],isOpen="",field=u"tradeDate,closePrice",pandas="1")
P_r2 = DataAPI.MktEqudGet(tradeDate=u"",secID=u"",ticker=u'600036',beginDate=date[0],endDate=date[-1],isOpen="",field=u"tradeDate,closePrice",pandas="1")

P_r1.index = list(P_r1.tradeDate)
P_r1['price'] = P_r1.ix[monthdate]['closePrice']
P_r1 = P_r1.dropna()

P_r2.index = list(P_r2.tradeDate)
P_r2['price'] = P_r2.ix[monthdate]['closePrice']
P_r2 = P_r2.dropna()

Return_P = pd.DataFrame()
Return_P['zglt'] = P_r1['price'].pct_change()-0.0025
Return_P['zsyh'] = P_r2['price'].pct_change()-0.0025
Return_P['opt'] = 0.261*Return_P['zglt']+0.739*Return_P['zsyh']
Return_P = Return_P.dropna()

choosedata = pd.concat([Return_P,Rturn,SMB,HML],axis = 1).dropna()
choosedata = choosedata.rename_axis({0:'SMB',1:'HML'}, axis='columns')
print choosedata
choosedata.to_csv(u'fundamental_data.csv')
```
