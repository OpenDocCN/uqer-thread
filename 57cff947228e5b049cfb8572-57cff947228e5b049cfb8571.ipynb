```python
import pandas as pd
import quartz_futures as qf
from quartz_futures.api import *
from quartz_futures.trade.cost import Slippage, Commission


# 策略回测参数，用于初始化回测环境。
benchmark = 'rb1610'                 # 策略参考基准
universe = ['rb1610']               # 策略证券池
start = pd.datetime(2016,7,1)   # 回测开始时间
end   = pd.datetime(2016,8,30)    # 回测结束时间
capital_base = 1e4                  # 初试可用资金
refresh_rate = 30                   # 调仓周期
freq = 'm'                          # 调仓频率：s -> 秒；m-> 分钟；d-> 日；
margin_rate = {'rb1610': 0.10}                           # 保证金比例，格式：合约代码 -> 比例
commission = Commission(cost=0.001, unit='perValue')     # 手续费
slippage = Slippage(value=0.0, unit='perValue')          # 滑点


### 策略初始化函数，一般用于设置计数器，回测辅助变量等。
def initialize(futures_account):
    pass

### 回测调仓逻辑，每个调仓周期运行一次，可在此函数内实现信号生产，生成调仓指令。
def handle_data(futures_account):
    future = futures_account.universe[0]
    data = futures_account.get_symbol_history(symbol = [future], time_range = 20)[future]
    if data.empty: return
    maxHigh = max(data.highPrice.head(-1))
    maxLast = max(data.closePrice.head(-1))
    minLast = min(data.closePrice.head(-1))
    minLow  = min(data.lowPrice.head(-1))
    bandRange = max(maxHigh-minLast,maxLast-minLow)
    buyBound = data.closePrice.values[0] + 0.5 * bandRange
    sellBound = data.closePrice.values[0] + 0.5 * bandRange
    if data.closePrice.values[-1] > buyBound and not futures_account.position:
        order(future, amount = 1, offset_flag = 'open', direction = 'long')
        order(future, amount = 1, offset_flag = 'close', direction = 'short')
    elif data.closePrice.values[-1] < sellBound and not futures_account.position:
        order(future, amount = 1, offset_flag = 'open', direction = 'short')
        order(future, amount = 1, offset_flag = 'close', direction = 'long')
    
bt, account = qf.backtest.backtest(universe=universe, start=start, end=end, 
                                   initialize=initialize, handle_data=handle_data, 
                                   capital_base=capital_base, refresh_rate=refresh_rate,
                                   freq=freq, margin_rate=margin_rate, commission = commission, slippage = slippage)

```

```python
import numpy as np
import matplotlib.pyplot as plt
from __future__ import division 
from matplotlib.font_manager import FontProperties
from matplotlib.ticker import MultipleLocator,FormatStrFormatter 

class performanceAnalysis:
    
    def __init__(self, bt = None, perf = None, **args):
        self.bt = bt
        self.perf = perf
        self.benchmark = args.get('benchmark',None)
        self.capitalBase = args.get('capital_base', 1e6)
        self.startDate = args.get('start',None)
        self.endDate = args.get('end', None)
        pass
            
    def naCurve(self):
        curve = ((np.mat(self.bt.portfolio_value)/self.capitalBase)*100).tolist()[0]
        fontsize, labelsize, linewidth = 15, 15, 2

        xmajorLocator = MultipleLocator(1)  
        ymajorFormatter = FormatStrFormatter('%.2f%%')

        plt.style.use('ggplot')
        fig = plt.figure(figsize=(12,8))
        
        ax = fig.add_subplot(1, 1, 1)
        ax.xaxis.set_major_locator(xmajorLocator)
        ax.yaxis.set_major_formatter(ymajorFormatter)     
        
        ax.spines['left'].set_color('grey')
        ax.spines['right'].set_color('grey')
        ax.spines['top'].set_color('grey')
        ax.spines['bottom'].set_color('grey')   

        with plt.style.context('fivethirtyeight'):
            plt.plot(curve,'r')
            plt.plot(self._benchmark(),'k')
        plt.grid(True)
        plt.legend(['N.A.','benchmark'],loc='best')
        plt.tick_params(labelsize=0.9*labelsize)
        plt.title('N.A.',fontsize = fontsize,verticalalignment = 'bottom',horizontalalignment = 'center')
        plt.xlabel('Time',fontsize=fontsize,verticalalignment = 'top', horizontalalignment = 'center')
        plt.ylabel('Profit', fontsize = fontsize,  verticalalignment = 'bottom', horizontalalignment = 'center',rotation=90)
        pass

    def performance(self):
        keys = ['total return','profit loss ratio','max drawdown', 'hit rate', 'action ratio']
        values = [
            str(round(self.perf['total_return']*100,2))+'%',
            round(self.perf['profit_loss_ratio'],2),
            str(round(abs(self.perf['max_drawdown'])*100,2))+'%',
            str(round(self.perf['hit_ratio']*100,2))+'%',
        ]
        perfDicts = dict(zip(keys, values))
        return pd.DataFrame(perfDicts,['performance'])
        pass
    
    def _benchmark(self):
        benchmark = DataAPI.MktFutdVolGet(ticker=self.benchmark,beginDate=self.startDate,endDate=self.endDate,field=u'closePrice',pandas="1")['closePrice'].tolist()
        benchmark = [round(100.0*index,2) for index in benchmark/benchmark[0]]
        return benchmark
        pass
        
    def _listStrip(self,list):
        tmp = []
        for index in list:
            tmp += index
        return tmp

pf = performanceAnalysis(bt, account,start = start, end = end, benchmark = universe[0], capital_base = capital_base)
pf.naCurve()
```
