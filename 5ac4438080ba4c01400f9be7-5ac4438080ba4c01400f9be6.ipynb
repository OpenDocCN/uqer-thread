```python
#用前复权P
import pandas as pd
import numpy as np
import matplotlib.pylab as pylab
from CAL.PyCAL import font
ct,gt = '600674','600886'
begin_date,end_date = '20130101','20150514'
ct_df = DataAPI.MktEqudAdjGet(ticker=ct,beginDate=begin_date,endDate=end_date,pandas="1")
gt_df = DataAPI.MktEqudAdjGet(ticker=gt,beginDate=begin_date,endDate=end_date,pandas="1")
liangtou_df = pd.DataFrame()
liangtou_df['川投股价'] = ct_df.closePrice
liangtou_df['国投股价'] = gt_df.closePrice
liangtou_df['川投国投比例'] = liangtou_df['川投股价']/liangtou_df['国投股价']
print '川投国投比例AVE:',np.mean(liangtou_df['川投国投比例'])
print '川投国投比例STD:',np.std(liangtou_df['川投国投比例'])
liangtou_df.plot( figsize=(16,10))
pylab.legend([u'川投股价',u'国投股价', u'川投国投股价比例'], prop=font)
```
