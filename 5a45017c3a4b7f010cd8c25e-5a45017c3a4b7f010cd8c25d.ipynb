# 4、解析部分成交、未成交订单及下单顺序对收益率的影响

直接用总资产乘以各股票权重计算购买股票手数并依此下单的话，会出现部分成交和未成交的情况，这是因为所需购买股票的资金不够，产生这种情况的原因有二：
	1、计算股票手数时未考虑佣金
    2、换仓时，停牌的股票无法卖出，占用部分资金
佣金问题很好解决，一般佣金不到1%，所以只需在总资产前乘以一个系数（0.99），留出部分资金用于支付佣金。
对于停牌的股票，我的处理方法是将这部分占用资金从总资产中扣除，然后用扣除后的总资产乘以股票权重计算手数。
#特别是在资金不足以全额购买股票，存在部分成交和未成交的情况下，即使是同一策略，下单顺序的不同会导致最终收益率的不同
#在quartz3下，account.universe应转换为context.get_universe(exclude_halt=True)

以下两段代码实现的都是【第16天：如何在优矿上一个人干掉一家公募量化团队！Alpha！Go！】中的FF三因子模型(FF模型明明有三个因子，示例代码却只包含两个因子，囧)，所不同的是第一段代码是我按照FF模型思想自己写的，第二段是示例代码按照Quartz3改的。因为代码的差异(交易股票池我是用的DateFrame，示例用的字典)，导致股票下单顺序不同，最终的收益率也不同。
