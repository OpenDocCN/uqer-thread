## 什么是短周期价量特征因子
短周期量价因子，属于市场情绪类因子，从股票的日线数据出发，根据墨菲的《股票/期货市场技术分析》所描述的股价波动规律，进一步构造技术指标衍生因子。比如：动量、反转、开盘缺口、量价背离、成交量放大等。技术分析流派认为某些特定形态描述了投资者的行为，可以甄别未来走势。短周期意味着，这类因子利用过去较短的日线数据构建，所以对未来短期1~5日有预测作用。研报中指出很多国外量化机构通过大量构造因子，来实现策略盈利。故此研报在worldquant-Alpha101基础上，扩充了大量短期因子。



## 短周期因子能带来哪些优势
短周期量价因子优势在于以下，首先，短周期因子可以弥补传统因子的短板，进一步解释特质收益，避免与其他财务、分析师因子的相关性。其次，短周期因子可以提高市场中策略的多样性，其适用于调仓频率高的交易者。最终，短周期因子对历史股价的特征挖掘，结合分析师对市场的理解，手工构造特征，更容易解释因子的物理含义

# 一、为何使用短周期交易型因子
交易型因子来源于量价数据，我们认为其中包含了交易情绪，而这些交易情绪可能会带来超额收益。经过实验，这类交易型因子与风格因子相关性较低，可以带来增量信息。使用短周期交易型因子有以下好处：
1) 超额收益来源：短周期交易型因子收益来源于交易行为，而不是股票的内在价值。
2) 因子数据来源：短周期交易型因子数据来源于日频量价数据，而不是财报、研报数据
3) 因子有效周期：短周期交易型因子适用周期一般在1周以内，而不是传统因子的月、季度
4) 策略交易频率：短周期交易型因子适合以T+1~T+5日换仓，适合短期交易者
6) 受市场风格影响：短周期交易型因子与市场风格相关性较低

## 3.1 因子体系的显著性检验及有效周期
这里通过回归单因子对下一期特质性收益的因子收益，判断是否有预测能力。
实验结果显示，T+2期调仓，使得平均年化收益率最高；但我们也发现单因子收益会出现负值，而研报中全为正，故考虑设置观察期，回归收益，添加因子方向


step1: 因子中性化(去除风格、行业影响), 已知因子取值 $X_{K}^t$, 取目标因子残差 $\epsilon_{K}^{t}$ (即优矿提供的neutralize函数)
       $\large X_{K}^t = \beta_{industry} X_{industry} + \beta_{style} X_{style} + \epsilon_{K}^{t}$

step2: 针对预测周期d, 回归单期因子收益率 $f_{k}$ (这里需要专业版，用风险模型的因子收益率，应该是d期的风格、行业收益)
       $\large R_{t+d} = f_{industry} X_{industry} + f_{style} X_{style} + f_{k} \epsilon_{K}^{t} + \epsilon_{t+d}$

step3: 计算因子收益率序列的年化收益 $E(f_{k})$ 及信息比率 $IR(f_{k})$，希望大于0且越大越好
       $\large E(f_{k}) = 252 \cdot (\overline{f_{k} / d})$
       $\large IR(f_{k}) = \frac{\sqrt{252} \cdot (\overline{f_{k} / d)}}{\sigma(f_{k} / d)}$

```python
from lib.gtja191 import *
import datetime
import pandas as pd
import numpy as np
import inspect
import statsmodels.formula.api as smf
import statsmodels.api as sm
from scipy import stats
import math
import cPickle as pickle
from IPython.display import display, HTML
from CAL.PyCAL import *
import matplotlib.pylab as plt
import quartz_extensions.Optimizer as opt
from copy import deepcopy
import scipy.optimize as sco
import sys
import traceback
import os.path
import multiprocessing as mp

past2pred = 250 # 使用过去250个交易日的因子收益移动平均作为下期预测值
methods = get_gtja_methods()
window_list = [i for i in range(1, 6)]
exclude_factors = ['alpha030', 'alpha026', 'alpha033', 'alpha039', 'alpha045', 'alpha056', 'alpha061', 'alpha092', \
                'alpha101', 'alpha108', 'alpha123', 'alpha125', 'alpha138', 'alpha149', 'alpha154', 'alpha184', \
                'alpha025', 'alpha075', 'alpha015', 'alpha017']

def load_pickle(pickle_name, ori_obj):
    if not os.path.exists(pickle_name):
        return False, ori_obj
    else:
        with open(pickle_name, 'rb') as handle:
            ori_obj = pickle.load(handle)
    return True, ori_obj

def save_pickle(pickle_name, ori_obj):
    with open(pickle_name, 'wb') as handle:
        pickle.dump(ori_obj, handle, protocol=pickle.HIGHEST_PROTOCOL)
    return

def get_trade_cal(start_date, end_date):
    trade_cal = DataAPI.TradeCalGet(exchangeCD=u"XSHG,XSHE",beginDate=start_date,endDate=end_date,field=u"calendarDate,isOpen,",pandas="1")
    trade_cal['calendarDate'] = trade_cal['calendarDate'].apply(lambda x: x.replace('-', ''))
    trade_cal = trade_cal[trade_cal['isOpen'] == 1].set_index('calendarDate', drop=True)
    return trade_cal

def load_daily_data(trade_cal, start_date, end_date, ticker='000300'):
    """
    一次性把起始日期内hs300的动态个股及依赖因子都加载到内存;
    仿Signal/SignalGenerator函数的输入格式，feed数据
    """
    all_tickers = []
    for trade_date in trade_cal.index[::max(len(trade_cal.index)/10, 10)]:
        cons_df = DataAPI.IdxConsGet(secID=u"",ticker=ticker,intoDate=trade_date,isNew=u"",field=u"consTickerSymbol,intoDate,outDate",pandas="1")
        all_tickers += list(cons_df['consTickerSymbol'].values)
    all_tickers = sorted(set(all_tickers))

    data = dict()
    # 1) 量价基本数据，这里建议按ticker one by one取
    factor_names = {'closePrice':'closePrice', 'highestPrice':'highPrice', 'lowestPrice':'lowPrice', 'openPrice':'openPrice', \
                    'turnoverValue':'turnoverValue', 'turnoverVol':'turnoverVol'}
    daily_k = pd.DataFrame()
    for ticker in all_tickers:
        tmp_df = DataAPI.MktEqudAdjGet(tradeDate="",secID=u"",ticker=ticker,isOpen="",beginDate=start_date,endDate=end_date,
                              field=','.join(["ticker,tradeDate"]+factor_names.keys()),pandas="1")
        daily_k = daily_k.append(tmp_df)
    daily_k['tradeDate'] = daily_k['tradeDate'].apply(lambda x: int(x.replace('-', '')))
    for factor_name in factor_names:
        data[factor_names[factor_name]] = pd.crosstab(index=daily_k['tradeDate'], columns=daily_k['ticker'], values=daily_k[factor_name], aggfunc='sum')
        # 由于面板数据cross出来nan值，学signal框架，我们forward-fill nan
        # data[factor_names[factor_name]] = data[factor_names[factor_name]].fillna(method='ffill')
    for window in window_list:
        data['ret_{0}'.format(window)] = data['closePrice'].pct_change(periods=window)
    print('[loadData][1/3] loaded price factor')
    
    # 2) 优矿已经计算的衍生因子
    factor_names = ['KDJ_J','PB','MktValue','VR','OBV6','BBIC','KDJ_K','BIAS5','CCI10','REVS20','KDJ_D','VSTD10','VSTD20','BBI','ADX','ADXR']
    factor_df = pd.DataFrame()
    part_trade_cal_index = trade_cal[(trade_cal.index>=start_date) & (trade_cal.index<=end_date)].index
    tickers = ','.join(all_tickers)
    for idx, trade_date in enumerate(part_trade_cal_index):
        daily_k = DataAPI.MktStockFactorsOneDayProGet(tradeDate=trade_date,secID=u"",ticker=tickers,field=','.join(["ticker,tradeDate"]+factor_names),pandas="1")
        daily_k['tradeDate'] = daily_k['tradeDate'].apply(lambda x: int(x.replace('-', '')))
        factor_df = factor_df.append(daily_k)
        if idx % 50 == 0:
            print('[loadData][2/3][{0}/{1}] loaded uqer factor'.format(idx, len(part_trade_cal_index)))
    for factor_name in factor_names:
        data[factor_name] = pd.crosstab(index=factor_df['tradeDate'], columns=factor_df['ticker'], values=factor_df[factor_name], aggfunc='sum')
    print('[loadData][2/3] loaded uqer factor')
    
    # 3) 个股收益-风险因子总收益=特质性收益，这里建议按ticker one by one取
    valid = False
    factor_spe = pd.DataFrame()
    start_date_hyphen = datetime.datetime.strptime(start_date, '%Y%m%d').strftime('%Y-%m-%d')
    end_date_hyphen = datetime.datetime.strptime(end_date, '%Y%m%d').strftime('%Y-%m-%d')
    while not valid:
        try:
            for ticker in all_tickers:
                tmp_df = DataAPI.RMSpecificRetDayGet(secID=u"",ticker=ticker,tradeDate="",beginDate=start_date_hyphen,endDate=end_date_hyphen,
                                               field="ticker,tradeDate,spret",pandas="1")
                factor_spe = factor_spe.append(tmp_df)
            factor_spe['tradeDate'] = factor_spe['tradeDate'].apply(lambda x: int(x.replace('-', '')))
            factor_spe = pd.crosstab(index=factor_spe['tradeDate'], columns=factor_spe['ticker'], values=factor_spe['spret'] / 100.0, aggfunc='sum')
            valid = True
        except Exception as e:
            # exc_type, exc_value, exc_traceback = sys.exc_info()
            # print(traceback.format_exception(exc_type, exc_value, exc_traceback))
            print('[loadData][3/3] load spe ret timeout')
    for window in range(1, 6):
        data['RM_RET_{0}'.format(window)] = data['ret_{0}'.format(window)] - \
            factor_spe.rolling(window=window,min_periods=window).apply(lambda x: np.cumprod(x+1)[-1]-1.0)
    print('[loadData][3/3] loaded risk ret data')
    return data


def comp_raw_neu_signal(signal, data, func, need_neu=False, sign_symbol=1):
    """
    根据func计算原始信号值;
    中性化放到的dict中;
    TODO: 按dynamic universe过滤
    """
    
    def construct_data(df, idx, max_window):
        """
        个股停牌情况做特殊处理；这里对每只个股取前N个正常交易的数据值。
        """
        tmp_df = pd.DataFrame(np.zeros((max_window, len(df.columns))), columns=df.columns)
        tmp_df[:] = np.nan
        for idy, ticker in enumerate(df.columns):
            values = df.iloc[:idx][ticker].dropna().iloc[-max_window:].values
            if len(values) == 0:
                continue
            tmp_df.iloc[-len(values):, idy] = values
        return tmp_dfs
    
    columns = inspect.getargspec(func).defaults[0]
    max_window = inspect.getargspec(func).defaults[1]
    func_name = func.__name__
    signal[func_name] = pd.DataFrame(index=data['ret_1'].index, columns=data['ret_1'].columns)
    for idx in range(max_window,len(signal[func_name].index)):
        td = signal[func_name].index.values[idx]
        tds = signal[func_name].index.values[idx-max_window:idx+1]
        # signal[func_name].loc[td] = func({col:data[col].loc[tds] for col in columns}).replace([np.inf, -np.inf], np.nan)
        signal[func_name].loc[td] = sign_symbol * func({col:construct_data(data[col], idx, max_window) for col in columns}).replace([np.inf, -np.inf], np.nan)
        x_not_na = signal[func_name].loc[td].dropna()
        if not need_neu or x_not_na.shape[0] == 0:
            continue
        x_not_na = (x_not_na.T).reset_index().rename(columns={'index':'ticker',td:'value'})
        series = x_not_na[['ticker','value']].set_index('ticker')['value'].dropna()
        try:
            neued = neutralize(standardize(winsorize(series)),target_date=str(td),exclude_style_list=[])
        except Exception as e:
            # exc_type, exc_value, exc_traceback = sys.exc_info()
            # print(traceback.format_exception(exc_type, exc_value, exc_traceback))
            print('[NeutralizeFactor] signal:{0} td:{1} len:{2} input not none is not enough'.format(func_name, td, len(series)))
            neued = series
        signal[func_name].loc[td] = (neued.to_frame().rename(columns={0:td}).T).iloc[-1]
        if idx % 100 == 0:
            print('[NeutralizeFactor][{0}/{1}] signal: {2} neued'.format(idx, len(signal[func_name].index), func_name))
    return signal


def comp_raw_signal(func, sign_symbol=1):
    """
    根据func计算原始信号值;
    中性化放到的dict中;
    TODO: 按dynamic universe过滤
    """
    global data
    def construct_data(df, idx, max_window):
        """
        个股停牌情况做特殊处理；这里对每只个股取前N个正常交易的数据值。
        """
        tmp_df = pd.DataFrame(np.zeros((max_window, len(df.columns))), columns=df.columns)
        tmp_df[:] = np.nan
        for idy, ticker in enumerate(df.columns):
            values = df.iloc[:idx][ticker].dropna().iloc[-max_window:].values
            if len(values) == 0:
                continue
            tmp_df.iloc[-len(values):, idy] = values
        return tmp_df
    
    process_name = mp.current_process().name
    columns = inspect.getargspec(func).defaults[0]
    max_window = inspect.getargspec(func).defaults[1]
    func_name = func.__name__
    res = pd.DataFrame(index=data['ret_1'].index, columns=data['ret_1'].columns)
    print('[RawFactor][{0}] process begin. max_window:{1} fun_name:{2}'.format(process_name, max_window, func_name))
    for idx in range(max_window,len(res.index)):
        t1 = time.time()
        td = res.index.values[idx]
        tds = res.index.values[idx-max_window:idx+1]
        # res.loc[td] = func({col:data[col].loc[tds] for col in columns}).replace([np.inf, -np.inf], np.nan)
        tmp_dict = {col:construct_data(data[col], idx, max_window) for col in columns}
        t2 = time.time()
        res.loc[td] = sign_symbol * func(tmp_dict).replace([np.inf, -np.inf], np.nan)
        t3 = time.time()
        if idx % 100 == 0:
            print('[RawFactor][{0}/{1}][{2}] signal:{3} prepare:{4:.2f} sig:{5:.2f}'.format(idx, len(res.index), process_name, func_name, t2-t1, t3-t2))
    return (func_name, max_window, res)


def comp_raw_signal_unpack(args):
    return comp_raw_signal(*args)


def comp_neu_signal(ori_signal, max_window):
    for idx in range(max_window,len(ori_signal.index)):
        t1 = time.time()
        td = ori_signal.index.values[idx]
        x_not_na = ori_signal.loc[td].dropna()
        if x_not_na.shape[0] == 0:
            continue
        x_not_na = (x_not_na.T).reset_index().rename(columns={'index':'ticker',td:'value'})
        series = x_not_na[['ticker','value']].set_index('ticker')['value'].dropna()
        try:
            neued = neutralize(standardize(winsorize(series)),target_date=str(td),exclude_style_list=[])
        except Exception as e:
            # exc_type, exc_value, exc_traceback = sys.exc_info()
            # print(traceback.format_exception(exc_type, exc_value, exc_traceback))
            print('[NeutralizeFactor] signal:{0} td:{1} len:{2} input not none is not enough'.format(func_name, td, len(series)))
            neued = series
        ori_signal.loc[td] = (neued.to_frame().rename(columns={0:td}).T).iloc[-1]
        if idx % 100 == 0:
            print('[NeutralizeFactor][{0}/{1}] signal:{2} neued'.format(idx, len(res.index), func_name))
    return ori_signal


def regress_single_factor_ret(signal, data, fun_name, window=1):
    """
    去除风格、行业影响，回归因子单期收益;
    TODO: 改为WLS回归
    """
    ys = data['ret_{0}'.format(window)].shift(-window)
    xs = signal[fun_name]
    cs = data['RM_RET_{0}'.format(window)].shift(-window)
    idxs = pd.Series(data=range(len(data['ret_1'].index)), index=data['ret_1'].index)
    def single_var_linregress(idx, ys, xs, cs):
        df = pd.concat([ys.iloc[idx].T - cs.iloc[idx].T, xs.iloc[idx].T], axis=1)
        df = df.dropna(axis=0)
        df.columns = ['y', 'x']
        if len(df.index) == 0:
            return np.nan
        try:
            res = stats.linregress(df['x'].values, df['y'].values)[0]
        except Exception as e:
            # exc_type, exc_value, exc_traceback = sys.exc_info()
            # print(traceback.format_exception(exc_type, exc_value, exc_traceback))
            print('[SingleFactorTest] signal:{0} regress ic, occurs zero division error'.format(func_name))
            return np.nan
        return res

    factor_ret = idxs.apply(lambda x: single_var_linregress(x, ys, xs, cs))
    factor_ret_annualized = 252 * np.mean((factor_ret / window).dropna())
    std_ret = np.std((factor_ret / window).dropna(), ddof=1)
    factor_ret_ir = math.sqrt(252) * np.mean((factor_ret / window).dropna()) / std_ret if std_ret != 0.0 else np.nan
    return factor_ret, factor_ret_annualized, factor_ret_ir


def generate_single_factor_summary(stats_res):
    """
    生成单因子测试报告
    """
    gb_res = stats_res.groupby(by=['window']).mean()
    print(gb_res.to_html())
    # 不同周期下因子收益率分布直方图
    f, axes = plt.subplots(5, sharex=True, sharey=True)
    for idx, window in enumerate(window_list):
        stats_res[stats_res['window'] == window]['ret_ann'].hist(ax=axes[idx], bins=20)
    axes[0].set_title('FR Annualized Average Histogram')
    plt.show()
    plt.close()
    f, axes = plt.subplots(5, sharex=True, sharey=True)
    for idx, window in enumerate(window_list):
        stats_res[stats_res['window'] == window]['ret_ir'].hist(ax=axes[idx], bins=20)
    axes[0].set_title('IR Annualized Average Histogram')
    plt.show()
    plt.close()
    # 不同周期下因子收益率均值及IR均值变化
    fig, ax = plt.subplots()
    ax2 = ax.twinx()
    ax.bar(window_list, gb_res['ret_ann'], width=0.3, color='b', label=u'FR Annualized Average')
    ax.legend(loc='upper left')
    ax2.plot([i+0.15 for i in window_list], gb_res['ret_ir'], 'c', linewidth=2.0, label=u'IR Annualized Average')
    ax2.legend(loc='upper right')
    ax.set_title('IR vs FR Average')
    plt.xticks(window_list)
    plt.show()
    plt.close()
```

## 3.2 因子合成后的预测能力
这里测试预测周期$t=1$的情况下，等权合成因子，利用过去250日的因子收益移动平均作为下一期预测值。通过判断因子收益预测值与真实值是否同号，判断胜率。
<br>
step1: 正交标准化, 已知因子取值 $X_{K}^t$, 取目标因子残差 $\epsilon_{K}^{t}$ (即优矿提供的neutralize函数)
       $\large X_{K}^t = \beta_{industry} X_{industry} + \beta_{style} X_{style} + \epsilon_{K}^{t}$

step2: 针对预测周期$t=1$, 回归过去时间段$T$内所有因子收益率簇 $\{f_{k}^{t}\}_{K}^T$

$\large R = f_{industry} X_{industry} + f_{style} X_{style} + f_{k1} \epsilon_{k1} + f_{k2} \epsilon_{k2} + ... + f_{K} \epsilon_{K} + \epsilon$

step3: 利用过去$T$时间段(这里与后文一致取250)因子收益率均值作为下期预测值，乘以当期每个目标因子残差截面，得到阿尔法收益截面预测向量，即
       $\large E(\epsilon_{t+1}) = \sum_{k=1}^{K}(\frac{\sum_{t=1}^{T} f_{k}^{t}} {T} \cdot \epsilon_{k}^{t})$
       
step4: 计算各期模型预测系数，进而计算模型系数$T$检验结果，其中$\epsilon_{t+1}$表示下一期末得到的真实阿尔法收益截面向量，即
       $\large \epsilon_{t+1} = \sum_{k=1}^{K} f_{k} \epsilon_{k}^{t+1}$
       $\large IC_{AlphaModel}^{t} = corr(E(\epsilon_{t+1}), \epsilon_{t+1})$
       $\large TStat_{AlphaModel} = \frac{\overline {IC}_{AlphaModel} \cdot \sqrt{T}}{\sigma(IC_AlphaModel)}$

```python
def regress_multi_factor_ret(signal, data, fun_names, window=1, past2pred=250):
    """
    去除风格、行业影响，回归多因子单期收益;
    计算每日IC,30日IC均值;x
    TODO: 改为WLS回归
    """
    ys = data['ret_{0}'.format(window)].shift(-window)
    cs = data['RM_RET_{0}'.format(window)].shift(-window)
    
    def multi_var_linregress(idx, ys, signal, fun_names, cs):
        lis = [ys.iloc[idx].T - cs.iloc[idx].T]
        lis += [signal[fun_name].iloc[idx].T for fun_name in fun_names]
        df = pd.concat(lis, axis=1)
        df.columns = ['y'] + fun_names
        df = df.dropna(axis=1, how='all')
        idx_cols = [idy for idy, col in enumerate(fun_names) if col in df.columns]
        res = [np.nan]*len(fun_names)
        df = df.dropna(axis=0)
        intersection_cols = [col for col in fun_names if col in df.columns]
        # print(df.head(5), idx_cols, df.columns)
        if len(df.index) == 0 or len(df.columns) <= 1 or 'y' not in df.columns:
            return res
        df = df.astype('float64')
        betas = sm.OLS(df['y'].values, df[intersection_cols].values, missing='drop').fit().params.tolist()
        for beta, idy in zip(betas, idx_cols):
            res[idy] = beta
        return res
    
    def comp_factor_ret_pred(idx, factor_rets_ma, signal, fun_names):
        lis = [signal[fun_name].iloc[idx].T for fun_name in fun_names]
        df = pd.concat(lis, axis=1)
        df.columns = fun_names
        df = df.multiply(np.squeeze(factor_rets_ma.iloc[idx].values), axis=1)
        df = df.sum(axis=1, skipna=True).T
        return df
    
    def comp_factor_ret_gt(idx, factor_rets, signal, fun_names):
        lis = [signal[fun_name].iloc[idx].T for fun_name in fun_names]
        df = pd.concat(lis, axis=1)
        df.columns = fun_names
        df = df.multiply(np.squeeze(factor_rets.iloc[idx].values), axis=1)
        df = df.sum(axis=1, skipna=True).T
        return df
    
    factor_rets = [multi_var_linregress(idx, ys, signal, fun_names, cs) for idx in range(len(data['ret_1'].index))]
    factor_rets = pd.DataFrame(factor_rets, index=data['ret_1'].index, columns=fun_names)
    factor_rets_ma = factor_rets.fillna(0.0).rolling(window=past2pred, min_periods=past2pred).mean()
    # print(factor_rets.iloc[250:])
    print('[MultiFactorTest] factor rets computed done')
    factor_rets_pred = [comp_factor_ret_pred(idx, factor_rets_ma, signal, fun_names) for idx in range(len(data['ret_1'].index))]
    factor_rets_pred = pd.DataFrame(factor_rets_pred, index=data['ret_1'].index, columns=data['ret_1'].columns)
    print('[MultiFactorTest] factor predict rets computed done')
    factor_rets_gt = [comp_factor_ret_gt(idx, factor_rets, signal, fun_names) for idx in range(len(data['ret_1'].index))]
    factor_rets_gt = pd.DataFrame(factor_rets_gt, index=data['ret_1'].index, columns=data['ret_1'].columns)
    print('[MultiFactorTest] factor groundtruth rets computed done')
    ic = factor_rets_pred.corrwith(factor_rets_gt, axis=1)
    ic30 = ic.rolling(window=30, min_periods=0).mean()
    ic_stats = pd.concat([ic, ic30], axis=1)
    ic_stats.columns = ['ic', 'ic30']
    print('[MultiFactorTest] ic computed done')
    return factor_rets_pred, ic_stats, factor_rets

def generate_multi_factor_summary(ic_stats, past2pred=250):
    """
    生成多因子合成后ic测试报告
    """
    ic_stats = ic_stats.dropna(axis=0)
    table1 = pd.DataFrame(data=[[np.mean(ic_stats['ic']), 
                       np.mean(ic_stats['ic']) * math.sqrt(past2pred) / np.std(ic_stats['ic'], ddof=1),
                       np.sum(ic_stats['ic'] > 0.0)/float(len(ic_stats.index))]], columns=['meanIC', 'Tstat', 'winPer'])
    print(table1.to_html())
    fig, ax = plt.subplots()
    ic_stats.index = ic_stats.index.astype(str)
    ic_stats['ic'].plot(ax=ax, color='b', linewidth=2.0, label='ic')
    ic_stats['ic30'].plot(ax=ax, color='r', linewidth=2.0, label='ic30')
    ax.legend()
    plt.show()
    plt.close()
    
def generate_multi_factor_num_summary(num_stats):
    # 因子个数表现表格
    print(num_stats.to_html())
    # 因子个数IC / Tstats表现
    num_stats = num_stats.T
    fig, ax = plt.subplots()
    ax2 = ax.twinx()
    ax.bar([i for i in num_stats.index], num_stats['Tstat'], width=5, color='b', label=u'T stats')
    ax.legend(loc='upper left')
    ax2.plot([i+5 for i in num_stats.index], num_stats['meanIC'], 'c', linewidth=2.0, label=u'mean IC')
    ax2.legend(loc='upper right')
    ax.set_title('mean IC vs T stats')
    plt.xticks(num_stats.index)
    plt.show()
    plt.close()
```

# 四、考虑换手率的组合构建
这里借鉴磊少的[优矿组合优化介绍](https://uqer.datayes.com/community/share/59b919dd8a86ca0111aaa63b)
结果看来，
1) 我们添加手续费约束后，对行业风险的控制并不能达到行业无风险暴露。
2) 回测结果，并未跑赢沪深300，可能是由于等权合成因子造成的。


研报中策略构建步骤，考虑手续费进行组合优化：
step1: 正交标准化, 已知因子取值 $X_{K}^t$, 取目标因子残差 $\epsilon_{K}^{t}$ (即优矿提供的neutralize函数)
       $\large X_{K}^t = \beta_{industry} X_{industry} + \beta_{style} X_{style} + \epsilon_{K}^{t}$
step2: <b>针对给定预测周期$t=2$</b>，回归过去时间段$T$内所有因子收益率簇 $\{f_{k}^{t}\}_{K}^T$

$\large R = f_{industry} X_{industry} + f_{style} X_{style} + f_{k1} \epsilon_{k1} + f_{k2} \epsilon_{k2} + ... + f_{K} \epsilon_{K} + \epsilon$
step3: 利用过去$T=250$日因子收益率均值作为下期预测值乘以当期每个目标因子残差截面，得到阿尔法收益截面预测向量，即
       $\large E(\epsilon_{t+1}) = \sum_{k=1}^{K}(\frac{\sum_{t=1}^{T} f_{k}^{t}} {T} \cdot \epsilon_{k}^{t})$
step4: 考虑换手手续费$Tc=0.003$、行业、风格中性为约束条件，求解组合权重$w$，即
       $\large \max{w \cdot E(\epsilon) - Tc \cdot \frac{\sum{|w_{t}-w_{t-1}|}}{2}}$
$\large \begin{aligned}s.t. & (w-w_{bm})X_{style} = 0 \\
       			            &(w-w_{bm})X_{industry} = 0 \\
                            & w \ge 0 \\
                            & \sum{w} = 1 \\
       \end{aligned}$

```python
def portfolio_optimize_uqer(date, prevdate, signal_df, prevweights_df=None):
    """
    调用uqer的优化器，进行组合构建;
    约束市值暴露与基准一致;
    约束行业暴露与基准一致;
    约束换手率不超过1.5倍(这里与研报稍微不一样,所以我们拿scipy再实现一次)
    这里signal的index需要为secID
    """
    # 前一次持仓权重
    if prevweights_df is None:
        prevweights_df = pd.DataFrame(np.ones((len(signal_df.index),1)), index=signal_df.index, columns=['weights']) / len(signal_df.index)
    prevweights = prevweights_df['weights'].fillna(0.0)
    
    signal = signal_df.dropna(axis=0)
    signal = signal['signal']
    if np.sum(np.isnan(signal.values)) > 0.2 * len(signal.index) or len(signal) == 0:
        print('[UqerOptimizer] nan signal above 20%')
        return False, prevweights_df
    

    # 风格约束,限制市值暴露
    size_bound = DataAPI.RMExposureDayGet(secID=",".join(signal.index), ticker="", tradeDate=prevdate, 
                                          field=['SIZE', 'secID']).set_index('secID').sum(axis=0) / len(signal)
    # 限制行业暴露
    pspec = opt.UqerOptimizer(signal=signal, construct_date=date, risk_model='day', benchmark_type='HS300', assets=prevweights)
    pspec.add_constraint(default_min_weight=0.0, default_max_weight=0.1)
    pspec.add_constraint(is_industry_neutralize=True)
    pspec.add_constraint(spec_style={'SIZE':size_bound})
    pspec.add_constraint(turnover_target=1.5)
    pspec.solve()
    
    assets_weights = pspec.assets[pspec.assets.optimal_weights > 0.00001][['optimal_weights']].rename(columns={'optimal_weights': 'weights'})
    return pspec.optimal, assets_weights

def portfolio_optimize_transaction(date, prevdate, signal_df, prevweights_df=None, disp=False):
    """
    简单用scipy实现研报中的组合构建
    https://docs.scipy.org/doc/scipy/reference/generated/scipy.optimize.minimize.html
    """    
    # 为了拿成份股权重
    begindate = datetime.datetime.strptime(date, "%Y%m%d").replace(day=1).strftime("%Y%m%d")
    nextmonth = datetime.datetime.strptime(date, "%Y%m%d").replace(day=28) + datetime.timedelta(days=4)
    enddate = (nextmonth - datetime.timedelta(days=nextmonth.day)).strftime("%Y%m%d")
    # print(prevdate, begindate, enddate)

    benchmark = DataAPI.IdxCloseWeightGet(secID=u"",ticker=u"000300", beginDate=begindate, endDate=enddate,
                                          field=u"consID,effDate,weight", pandas="1") \
                    .rename(columns={'consID':'secID'}) \
                    .sort_values(by=['secID', 'effDate'], ascending=[True, False]) \
                    .drop_duplicates(subset=['secID'], keep='first').set_index('secID', drop=True).drop('effDate', 1)
    benchmark['weight'] = benchmark['weight'] / np.sum(benchmark['weight'])

    # 风险因子暴露,约束风格因子时会用到
    factors = ['BETA','MOMENTUM','SIZE','EARNYILD','RESVOL','GROWTH','BTOP','LEVERAGE','LIQUIDTY','SIZENL',
               'Bank','RealEstate','Health','Transportation','Mining','NonFerMetal','HouseApp','LeiService','MachiEquip',
               'BuildDeco','CommeTrade','CONMAT','Auto','Textile','FoodBever','Electronics','Computer','LightIndus','Utilities',
               'Telecom','AgriForest','CHEM','Media','IronSteel','NonBankFinan','ELECEQP','AERODEF','Conglomerates']
    exposure = DataAPI.RMExposureDayGet(secID=",".join(benchmark.index.values), ticker="", tradeDate=prevdate,
                                        field=','.join(['secID'] + factors)).set_index('secID', drop=True)

    if prevweights_df is None:
        prevweights_df = pd.DataFrame(np.ones((len(benchmark.index),1)), index=benchmark.index, columns=['weights']) / len(benchmark.index)
        
    if np.sum(np.isnan(signal_df['signal'].values)) > 0.2 * len(signal_df.index):
        print('[ScipyOptimizer] nan signal above 20%')
        return False, prevweights_df

    exposure = exposure.merge(benchmark, left_index=True, right_index=True, how='outer')
    exposure = exposure.merge(signal_df, left_index=True, right_index=True, how='outer')
    exposure = exposure.merge(prevweights_df, left_index=True, right_index=True, how='outer')
    exposure['weights'] = exposure['weights'].fillna(0.0)
    exposure = exposure.dropna(axis=0)
    # print(exposure.head(5))
    exposure_factors = exposure[factors]
    exposure_constraints = exposure_factors.multiply(np.squeeze(exposure['weight'].values), axis=0).sum(axis=0)
    signal = exposure['signal']
    prevweights = exposure['weights']
    n_assets = len(signal.index)
    weights = pd.DataFrame(np.ones((n_assets,1)), index=exposure.index, columns=['weights']) / n_assets

    def optimize_transaction(weights, prevweights, signal):
        loss = np.sum(weights * signal) - 0.0015 * np.sum(abs(weights - prevweights))
        return -loss
    
    def optimize_transaction_jac(weights, prevweights, signal):
        jac = signal - np.sign(weights)
        return -jac

    cons = ({'type': 'eq', 
             'fun': lambda x: np.sum(x)-1,
             'jac' : lambda x: np.ones(len(x))}, 
            {'type': 'ineq',
             'fun': lambda x: 0.03 - np.sum(abs(exposure_factors.multiply(x, axis=0).sum(axis=0)-exposure_constraints)),
             'jac': lambda x: np.ones(len(x))})
    bnds = tuple((0,1) for x in range(n_assets))
    opts = sco.minimize(fun=optimize_transaction, jac=optimize_transaction_jac, x0=weights.values, args=(prevweights, signal,), method='SLSQP', bounds=bnds, constraints=cons, tol=1e-5, options={'maxiter':300, 'disp':disp})
    
    weights[:] = opts.x.round(3).reshape((n_assets, 1))
    return opts.success, weights

def validate_optimizer_results(date, asset_weights):
    """
    参见专业版，磊少的《优矿组合优化器使用文档》
    """
    from CAL.PyCAL import *
    begindate = datetime.datetime.strptime(date, "%Y%m%d").replace(day=1).strftime("%Y%m%d")
    nextmonth = datetime.datetime.strptime(date, "%Y%m%d").replace(day=28) + datetime.timedelta(days=4)
    enddate = (nextmonth - datetime.timedelta(days=nextmonth.day)).strftime("%Y%m%d")
    # 获得基准成分股行业分类
    equ_indu = DataAPI.EquIndustryGet(industryVersionCD=u"010303", secID=set_universe('HS300', date), intoDate=date, field=u"secID,industryName1", pandas="1")
    benchmark_equ_weight = DataAPI.IdxCloseWeightGet(ticker='000300', beginDate=begindate, endDate=enddate, field=u"consID,weight",pandas="1")
    benchmark_equ_weight.rename(columns={'consID': 'secID', 'weight':u'基准行业权重'}, inplace=True)
    # 获得基准行业权重
    benchmark_indu_weight =equ_indu.merge(benchmark_equ_weight, on='secID', how='inner').groupby(by='industryName1').sum()/100
    equ_indu_ = DataAPI.EquIndustryGet(industryVersionCD=u"010303", secID=list(asset_weights.index), intoDate=date, field=u"secID,industryName1", pandas="1")
    # 优化器行业权重
    pspec_indu_weight=equ_indu_.merge(asset_weights[['weights']], left_on='secID', right_index=True, how='inner').groupby(by='industryName1').sum()
    pspec_indu_weight.rename(columns={'optimal_weights':u'组合行业权重'}, inplace=True)
    data = pspec_indu_weight.merge(benchmark_indu_weight, left_index=True, right_index=True, how='inner')
    fig = plt.figure(figsize=(16,5))
    ax = fig.add_subplot(111)
    ax = data.plot(kind='bar', ax=ax)
    labels = [label.decode("utf-8") for label in data.index.values] 
    ax.set_xticklabels(labels, fontproperties=font, fontsize=13)
    ax.legend(prop=font)
    ax.set_xlabel(u'行业名称', fontproperties=font, fontsize=15)
    ax.set_ylabel(u'权重', fontproperties=font, fontsize=15)
    ax.grid()
    return
```
