## 概率统计方法：stats模块

概率分布
连续分布中包括
1. 均匀分布
2. 正态分布
3. t分布

离散分布中包括
1. 伯努利分布
2. 几何分布


```python
# 以正态分布为例
from scipy.stats import norm
import numpy as np
from matplotlib import pylab # 为了画图展示方便

method = dir(norm)
method = [x for x in method if (not '_' in x)]
print method
```

```python
# 随机生产5000个点，满足正态分布
x_norm = norm.rvs(size=5000, loc=0.5, scale=0.5)
# x_norm = np.random.normal(size=500, loc=0.5, scale=0.5)
```

```python
# 在这组数据下，用极大似然估计正态分布的参数
x_mean, x_std = norm.fit(x_norm)
print 'mean: %s, std: %s' % (x_mean, x_std)
```

```python
# 直方图及概率密度函数的展示
h = pylab.hist(x_norm, normed=True, bins=20)
x = np.linspace(-3, 3, 50)
p = pylab.plot(x, norm.pdf(x, loc=x_mean, scale=x_std), 'r-')
```

通过对pdf进行积分可以求解落在某个区间的概率大小，我们导入积分模块

```python
from scipy.integrate import trapz
x1 = np.linspace(-1,1,100)
p = trapz(norm.pdf(x1, loc=x_mean, scale=x_std), x1) 
print '{:.2%} of the values lie between -1 and 1'.format(p)

pylab.fill_between(x1, norm.pdf(x1, loc=x_mean, scale=x_std), color = 'red')
pylab.plot(x, norm.pdf(x, loc=x_mean, scale=x_std), 'k-')
```

# 优化方法：optimize模块

```python
from scipy import optimize
```

定义函数 $f(x) = x^2 + 10 x + 10$，
1. 求解该函数的根
利用求根公式可以得到两个解
2. 求解该函数的最小值
对$f(x)$求导得到$f'(x)=2x+10$，另$f'(x)=0$得到$x=-5$

```python
# 定义函数
def f(x):
    return x ** 2 + 10 * x + 10
```

```python
x = np.arange(-10, 0, 0.1)
pylab.plot(x, f(x))
```

```python
# 求解f(x)的根
print optimize.fsolve(f, 0) # 根1
print optimize.fsolve(f, -7) # 根2
```

```python
print optimize.fsolve(f, -5)
```

```python
print optimize.fsolve(f, -5, fprime= lambda x: 2 * x + 10) # 根2
```

```python
optimize.fmin_bfgs(f, 0)
```

```python
optimize.minimize(f, 0).x
```

```python
cons = ({'type': 'ineq', 'fun': lambda x: x + 4})
```

```python
optimize.minimize(f, 0, constraints=cons).x
optimize.mi
```

# 作业

定义函数
$f(x, y) = 2xy + 2x - x^2 - 2y^2$
带有约束条件
1. $2x - y = 0$
2. $y - 1 \geq 0$

求解$f(x, y)$的最大值

```python
def f(x):
    return -(2 * x[0] * x[1] + 2 * x[0] - x[0] ** 2 - 2 * x[1] ** 2)
```

```python
cons = (
    {'type': 'eq', 'fun': lambda x: 2 * x[0] - x[1]},
    {'type': 'ineq', 'fun': lambda x: x[1] - 1}
)
```

```python
x0 = np.array([0, 0])
x = optimize.minimize(f, x0, method='SLSQP', constraints=cons).x
print x
```
