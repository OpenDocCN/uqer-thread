### 下面是统计时间：行业轮动是否有动量效应，这个月排名靠前的行业下月维持的几率有多大呢？###

```python
#当月排名前threshold的下月继续的几率
def rank_chance(df,threshold,month):
    count=0
    month_count=len(df.index)
    total=(month_count-1)*threshold
    for i in range(0,len(df_month_rank)-1):
        l_now=list(df.iloc[i,:].sort_values(ascending=False)[:threshold].index)
        l_next=list(df.iloc[i+1,:].sort_values(ascending=False)[:threshold].index)
        count+=len(set(l_now).intersection(set(l_next)))
    return count,total,float(count)/int(total)
```

```python
#前N月排名前threshold的下月继续的几率
def rank_chance_N(df,threshold,N):
    total_count=0
    count=0
    for i in range(N-1,len(df_month_rank)-1):
        #取得当月前N位行业
        l_now=list(df.iloc[i,:].sort_values(ascending=False)[:threshold].index)
        for j in range(i-(N-1),i):
            #首先计算前N月都出现的行业
            l_pre=list(df.iloc[j,:].sort_values(ascending=False)[:threshold].index)
            l_now=list(set(l_now).intersection(set(l_pre)))
        total_count+=len(l_now)
        l_next=list(df.iloc[i+1,:].sort_values(ascending=False)[:threshold].index)
        count+=len(set(l_now).intersection(set(l_next)))
        if total_count>0:
            result=float(count)/int(total_count)
        else:
            result=0
    return count,total_count,result
```

先看排列本月排名前三分之一的行业下月继续在前三分之一的几率

```python
df_month_rank=df_month_rank.T
df_month_rank.drop(['平均','最高','最低'],axis=0,inplace=True)
df_month_rank.sort_index(ascending=True,inplace=True)
count,total,result=rank_chance(df_month_rank,6,1)
print(count,total,result)
```

只有32%，再看排名前二分之一的

```python
count1,total1,result1=rank_chance(df_month_rank,9,1)
print(count1,total1,result1)
```

**进一步，如果是前N个月排名前列的呢？**

```python
result={}
totals={}
#计算连续month_count月排名前threshold位，下月继续排名前threshold位的几率。
#df_result的index格式：threshold-month_count
for threshold in [3,6,9]:
    for month_count in range(2,6):
        count,total,r=rank_chance_N(df_month_rank,threshold,month_count)
        result[str(threshold)+'-'+str(month_count)]=r
        totals[str(threshold)+'-'+str(month_count)]=total
df_results=pd.DataFrame([result,totals]).T
df_results.columns=['几率','总数']
df_results
```

排名前二分之一的连续出现2-3个月的机会更大，而且排名继续在前二分之一的几率也大。
那么测试一下，如果每月等权买入前N月(N=1-5)排名前threashold的指数（由于ETF不完善，这里模拟买入指数），结果如何：

```python
data=exceed_month.drop('沪深300',axis=1)
data.dropna(inplace=True)
data.sort_index(ascending=True,inplace=True)
```

```python
#模拟等权买入前N月排列前threashold位的指数
def simu_buy(df,threshold,N,asc):
    from datetime import *
    result={}
    for i in range(1,len(df)):
        increase_now=1#本月涨幅
        if N==1:
            #取上月涨幅前列的行业
            l_now=list(df.iloc[i-1,:].sort_values(ascending=asc)[:threshold].index)
            #取当月涨幅均值
            increase_now=df.loc[df.index[i],l_now].mean()
        else:
            if i<=N:
                increase_now=1
            else:
                #取得上月前N位行业
                l_now=list(df.iloc[i-1,:].sort_values(ascending=asc)[:threshold].index)
                for j in range(i-N-1,i-1):
                    #首先计算前N月都出现的行业
                    l_pre=list(df.iloc[j,:].sort_values(ascending=asc)[:threshold].index)
                    l_now=list(set(l_now).intersection(set(l_pre)))
                if len(l_now)==0:
                    increase_now=1
                else:
                    #print(i,df.loc[df.index[i],l_now])
                    increase_now=df.loc[df.index[i],l_now].mean()
        result[datetime.strftime(df.index[i],'%Y-%m-%d')]=increase_now
        s_result=pd.Series(result).cumprod()
    return s_result
```

```python
results={}
asc=False
for threshold in [6,9]:
    for month_count in range(1,5):
        results[str(threshold)+'-'+str(month_count)]=simu_buy(data,threshold,month_count,asc)
Momentum=pd.DataFrame(results)
```

看上去收益不怎样？注意这是在熊市期间，**同期沪深300最终收益是0.72**，这几种参数只有连续4个月前二分之一的组合跑输了大盘。

试了动量策略，再看看反转策略怎么样？
这次的测试条件是，买入连续N个月涨幅后threshold位的指数结果如何？

```python
results={}
asc=True
for threshold in [6,9]:
    for month_count in range(1,5):
        results[str(threshold)+'-'+str(month_count)]=simu_buy(data,threshold,month_count,asc)
reverse=pd.DataFrame(results)
```

```python
df_mom=pd.DataFrame([Momentum.iloc[-1],reverse.iloc[-1]]).T
df_mom.columns=['动量','反转']
df_mom
```
