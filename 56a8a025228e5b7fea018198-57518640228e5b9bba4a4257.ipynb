# 大师系列13--罗伯·瑞克超额现金流选股法则

## 罗伯·瑞克亦是价值投资的拥护者，其于 1983 年成立公司后，就研发出了以公司自由现金流为选股指标的投资策略，并且提出股价/自由现金流作为评价投资标的风险的重要指标。
--------

>

#### 结论
#### 1、本策略表现尚可，可进行后续优化。
#### 2、由于选股条件苛刻，每期满足条件个股数量很少，可以适当放宽指标要求。
#### 3、对比优化前后的策略收益，大概率以下说法成立：本策略超额收益来源于精准选股，因此调仓频率适中，控制个股数量可以增加年化收益。
