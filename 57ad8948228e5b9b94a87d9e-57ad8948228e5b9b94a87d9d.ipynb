**Alpha#27**
$$ ((0.5 < rank((sum(correlation(rank(volume), rank(vwap), 6), 2) / 2.0))) ? (-1 * 1) : 1)$$



## **Tips**

**Rank(x)** ： 排列所有工具的 x 值，然后返回的数值是介于0.0和1.0之间均匀分布的浮点数。<br>
例如假设你有一个向量：<br>
A = [29, 12, 5, 7, 10]<br>
现在，将会为向量里的每个元素**按升序分级**，最小一级为 0：<br>
A = [29, 12, 5, 7, 10]<br>
A' = [4, 3, 0, 1, 2] <br>
每个“等级”再除以（元素个数 - 1）。于是得到：<br>
Rank (A) = [1.00, 0.75, 0.00, 0.25, 0.50]<br>
所以您等到了一个从 0 到 1 间距相等的向量。<br>

**Correlation(x, y, n)** : 	计算向量 x 中的值在过去n天的相关系数。请注意，n 必须小于256。

## **公式分析**

观察公式发现，公式涉及的数据有成交量volume和成交量加权价格vwap。 而且其中计算了两者的correlation， 看来，又是基于**量价关系**的alpha。 

该公式的逻辑是： 计算每只股票的rank((sum(correlation(rank(volume), rank(vwap), 6), 2) / 2.0)) - 0.5 ，如果大于0，赋予值-1（做空）；否则赋值 1（做多）。 也就是，我们买入的是rank()后值小的股票。也就是相关系数均值小的股票，**也就是负相关强的股票！**（这里并不是直接的量价负相关，这是因为rank的影响，在Alpha#03的帖子评论中有讨论过，会有一些差异。）

这与之前写过的[Zura论文附录中WorldQuant的Alpha#03](https://uqer.datayes.com/community/share/5750f4a1228e5b86a61fd7dc)中的alpha 03:$$ Alpha 03：
(−1∗correlation(rank(open),rank(volume),10))$$ 很相似。但也有些差异：

1.计算了前6天的rank(量)、rank(价)的相关系数，并取最近的两个相关系数取均值。

2.rank()-0.5的处理有所不同：03相当于买入的都是负相关的股票，但这里是买入**负相关排名前一半的股票，也就是有可能买入正相关的股票，也有可能没有全部买入负相关的股票**

3.等权值处理，而03是根据负相关的相关系数加权的

4.数据不同，这里的“价”为vwap，比open多了成交量的信息。

综合来看，该alpha比03更精致~ 下面来看看效果

由于signal模块暂时没有加入成交额数据，我们自己取数据来计算。计算时间略长~

设置股票池全A，最后选权重前100的股票买入。（其实是等权的。。前100后100理论上差距应该不大，不过测了下还真有些差距呢~）
