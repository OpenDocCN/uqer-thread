从股价红利比率到资产估值：
假设每年的分红是1块，然后分到无穷无尽，但是现在的无风险利率是5%，那么你的这笔永续的分红在现在值多少？

其实就是1+1/1.05+1/1.05^2...+1/1.05^24...

等比数列求和或者一个近似编程范例

```python
sum = 0
for i in range(10000):
    sum = sum + 1/(1.05 ** i)
print(sum)
```

实际上，稳定分红的现值计算是，divCash / (1-q)，其中q = 1 / (1 + 无风险利率)。如果有人硬要特别严谨的说的话，这里的无风险利率实际上可以解释成，无风险利率+通货膨胀率。但是简单的拿个债券基金当作无风险利率，对我而言胜过找到通货膨胀率。此外还有个divCash被收税的问题，总之，别管这些罗里吧嗦了。

对于比率差距从56到18的我的臆断的解释就是，就是A股有人习惯性粗暴的从市值看傻大粗，认为没有成长性。所以上面的比率，大的会比小的低。

为何固定分红是那么适用的一个模型。股权集中属于国家，在这个时候，如果不通过分红，很难有资本退出的渠道；而且它们普遍比较大，为市场所不喜。

大致的说，想看银行股高估不高估，一个简易的方法是，年度分红 * 20。这个的前提是差不多5%的无风险利率，并且，由于分红的折现也是有风险的，可能将来多了呢，可能将来少了呢。所以可以再稍稍的除以一个风险调整因子，这个风险调整因子具体值不知道。唯有在你选定20或者21的时候，加上风险，应该要比这个算出来的值小。

实际上上面的一个无聊代码也表明了，你对无风险利率的预期越小，那么你对银行股的预估值应该越高……当你预估无风险利率是4%的时候，后面的乘数20要变成26.

为何固定每年发放的红利，主要是看了建行、中国银行等等的红利分配，股息比股市稳定多了。说个题外话，如果这些大银行适当增加分红率，我是觉得股票价格会涨，反而不是跌……

对小股东而言，只有股东能够拿到手的才是真的，放在公司里的资产，大股东的一言堂，那么我觉得惰性当然会很大。这么多年稳定的分红下来，堪比债市的银行股。市场预期已经很稳定了。
