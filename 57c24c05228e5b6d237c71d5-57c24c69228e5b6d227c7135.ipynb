在上两周写的的基本面因子策略和风格轮动选股策略之后，本周写了一个简单的统计套利策略。统计套利策略，顾名思义，即根据股票的历史市场数据，获取相似度最高的两只股票形成一对配对股票，从而进行配对交易。所谓配对交易，是基于股价长期均衡模型的，即股票的价格最终会回归至其均衡价值。因此，当个股股价远远高于或者低于这一均衡价值时，我们便可以进行套利。而这一均衡价值恰恰是通过配对股票间的均衡关系度量的。

本策略选用二分K-means方法，通过观察期（前三年）内的历史市场数据（收盘价）对股票进行分类，并以同一类中的两只股票作为配对股票A、B。之后，采用Support Vector Regression（SVR）方法来拟合配对股票观察期内的历史市场数据，确定两者的均衡关系及其相应参数。在确立均衡关系之后，选取观察期后的三个月作为交易期，进行配对交易。

具体的配对交易原则为：
建仓：当A股票股价高于其基于B股票股价的均衡价值1倍标准差以上时，做多B股票并做空相同价值的A股票；当A股票股价低于其基于B股票股价的均衡价值一倍标准差以上时，做多A股票，并做空相同价值的B股票 。
平仓：当A股票与其基于B股票股价的均衡价值的差距基本为0时，执行平仓操作。
强制平仓：当A股票股价与其基于B股票股价的均衡价值差距3倍标准差以上时，执行强制平仓操作。

但由于我国A股市场对于融资融券的限制，很难直接进行做空。因此，在回测中我们使用修改后的配对交易原则：
1.初次建仓：当A股票股价高于其基于B股票股价的均衡价值1倍标准差以上时，买入B股票；当A股票股价低于其基于B股票股价的均衡价值一倍标准差以上时，买入A股票。
2.调仓操作A：当A股票股价高于其基于B股票股价的均衡价值1倍标准差以上时，若A股票的持仓数量大于0，则卖出A股票并买入相同价值的B股票；
 调仓操作B：当A股票股价低于其基于B股票股价的均衡价值一倍标准差以上时，若B股票的持仓数量大于0，则卖出B股票并买入相同价值的A股票 。
3.平仓：交易期结束时平仓。

完成上述数据处理方式和交易原则的确定后，我们对该策略进行回测。回测参数设置如下：以2013年1月1日至2015年12月31日的三年时间作为观察期，以2016年1月1日至2016年3月31日作为交易期，证券池为申万行业分类中的石油化工行业所有股票，参考标准为上证综指，调仓周期为1日。在回测中，我们通过二分K-means和SVR方法，在证券池中获取了上海石化（600688）和华联控股（000036）作为配对股票，并通过价格预测图展示了其均衡关系。依据其均衡关系，在交易期对其进行配对交易。最终的回测结果显示，本策略在收益上取得了不俗的表现，其年化收益预计能够达到34%，远高于同期上证综指的-51%。

然而，改动后的配对交易原则并不完全合理，因为其不仅需要一定的初始资金，也不能通过强制平仓操作来很好地规避风险。本策略同样给出了在允许融资融券下的配对交易原则相应代码，有兴趣的quant可以尝试用其在其它平台进行试验，回测效果将会更佳。

```python
from numpy import *
import pandas as pd

# Define the method of K-means
def distEclud(vecA, vecB):
    return sqrt(sum(power(vecA - vecB, 2))) #la.norm(vecA-vecB)
    
def randCent(dataSet, k):
    n = shape(dataSet)[1]
    centroids = mat(zeros((k,n)))#create centroid mat
    for j in range(n):#create random cluster centers, within bounds of each dimension
        minJ = min(dataSet[:,j])
        rangeJ = max(dataSet[:,j]) - minJ
        centroids[:,j] = mat(minJ + rangeJ[0,0] * random.rand(k,1))
    return centroids

def kMeans(dataSet, k, distMeas=distEclud, createCent=randCent):
    m = shape(dataSet)[0]
    clusterAssment = mat(zeros((m,2)))#create mat to assign data points 
                                      #to a centroid, also holds SE of each point
    centroids = createCent(dataSet, k)
    clusterChanged = True
    while clusterChanged:
        clusterChanged = False
        for i in range(m):#for each data point assign it to the closest centroid
            minDist = inf; minIndex = -1
            for j in range(k):
                distJI = distMeas(centroids[j,:],dataSet[i,:])
                if distJI < minDist:
                    minDist = distJI; minIndex = j
            if clusterAssment[i,0] != minIndex: clusterChanged = True
            clusterAssment[i,:] = minIndex,minDist**2
        for cent in range(k):#recalculate centroids
            ptsInClust = dataSet[nonzero(clusterAssment[:,0]==cent)[0]][0]#get all the point in this cluster
            centroids[cent,:] = mean(ptsInClust, axis=0) #assign centroid to mean 
    return centroids, clusterAssment

def biKmeans(dataSet, k, distMeas=distEclud):
    m = shape(dataSet)[0]
    clusterAssment = mat(zeros((m,2)))
    centroid0 = mean(dataSet, axis=0).tolist()[0]
    centList =[centroid0] #create a list with one centroid
    for j in range(m):#calc initial Error
        clusterAssment[j,1] = distMeas(mat(centroid0), dataSet[j,:])**2
    while (len(centList) < k):
        lowestSSE = inf
        for i in range(len(centList)):
            ptsInCurrCluster = dataSet[nonzero(clusterAssment[:,0]==i)[0],:][0]#get the data points currently in cluster i
            centroidMat, splitClustAss = kMeans(ptsInCurrCluster, 2, distMeas)
            sseSplit = sum(splitClustAss[:,1])#compare the SSE to the currrent minimum
            sseNotSplit = sum(clusterAssment[nonzero(clusterAssment[:,0]!=i)[0],1])
            if (sseSplit + sseNotSplit) < lowestSSE:
                bestCentToSplit = i
                bestNewCents = centroidMat
                bestClustAss = splitClustAss.copy()
                lowestSSE = sseSplit + sseNotSplit
        bestClustAss[nonzero(bestClustAss[:,0] == 1)[0],0] = len(centList) #change 1 to 3,4, or whatever
        bestClustAss[nonzero(bestClustAss[:,0] == 0)[0],0] = bestCentToSplit
        centList[bestCentToSplit] = bestNewCents[0,:].tolist()[0]#replace a centroid with two best centroids 
        centList.append(bestNewCents[1,:].tolist()[0])
        clusterAssment[nonzero(clusterAssment[:,0] == bestCentToSplit)[0],:]= bestClustAss#reassign new clusters, and SSE
    return mat(centList), clusterAssment
```

```python
# Define the security pool. Here we use the stock pool of Shiyou Huagong industry of Shenwan industry classfication.
universe=set_universe(IndSW.ShiYouHuaGongL2)  

# Get the market data between 01/01/2013 and 31/12/2015
prc_obs=DataAPI.MktEqudAdjGet(secID=universe,beginDate="20130101",endDate="20151231",field=["secID","tradeDate","closePrice"],pandas="1")

# Reform the data 
prc_merge=list(prc_obs.groupby('secID'))[0][1][['tradeDate','closePrice']]
for i in range(1,shape(list(prc_obs.groupby('secID')))[0]):
    prc_merge=pd.merge(prc_merge,list(prc_obs.groupby('secID'))[i][1][['tradeDate','closePrice']],on='tradeDate',how='outer')
columns=['tradeDate']
for i in range(0,shape(list(prc_obs.groupby('secID')))[0]):
    columns.append(list(prc_obs.groupby('secID'))[i][0])
prc_merge.columns=columns
```

```python
# Delete the stock data with NA values (Not Available)
prc_merge.dropna(axis=1,inplace=True)

# Use the K-means method to classfity the stocks and return a pair of stocks in the same class. 
prc_mat=mat(prc_merge.drop('tradeDate',axis=1)).T
label_mat=biKmeans(prc_mat, shape(prc_mat)[0]-1, distMeas=distEclud)[1]
label={}
j=0
for secID in prc_merge.columns[1:]:
    label[secID]=label_mat[j,0]
    j+=1
    
for i in label:
    for j in label:
        if (i!=j) & (label[i]==label[j]):
            universe=[i,j]
print('配对股票为 %s' % universe)
```

```python
# Calculate the corralation of the stock pair.
print('配对股票相关系数为 %s' % prc_merge.drop('tradeDate',axis=1).corr()[universe[0]][universe[1]])
```

```python
# Use the close price data of stock 0 to predict the close price of Stock 1 by Support Vector Regression (rbf and linear)
import matplotlib.pyplot as plt
from sklearn import svm

# Reform the price data to get the input X and target y
X=[]
for i in prc_merge[universe[0]]:
    X.append([i])
y=prc_merge[universe[1]]

# Fit Support Vector Regression models (rbf and linear)
svr_rbf=svm.SVR(kernel='rbf', C=1e3, gamma=0.1).fit(X, y)
svr_lin = svm.SVR(kernel='linear', C=1e3).fit(X, y)
y_rbf=svr_rbf.predict(X)
y_lin=svr_lin.predict(X)

# Visualize the regression results
plt.scatter(X, y, c='k', label='data')
plt.plot(X, y_rbf, c='g', label='RBF model')
plt.plot(X, y_lin, c='r', label='Linear model')
plt.xlabel('Close Price of Stock 600688')
plt.ylabel('Close Price of Stock 000036')
plt.title('Price Prediction by Support Vector Regression')
plt.legend()
plt.show() 
```
