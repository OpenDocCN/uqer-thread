### 二次规划问题的标准形式
$$min \quad \frac{1}{2}x^TPx+q^Tx$$
$$s.t. \quad Gx\leq h $$
$$\qquad Ax=b$$
$\qquad$ 上式中，$x$为所要求解的列向量,$x^T$表示$x$的转置

# 在计算马科维兹有效前沿情景中，在特定目标收益的前提下，找出方差最小的权重向量w，也就是这里的$x$
#### 1，目标函数中P、q的值
P是协方差矩阵；这里的协方差矩阵是日度收益率，计算出来的方差也是基于日度；如果要获取年化方差，需要乘以250；为了计算方便，在初始计算日度协方差时，乘以250
q是一维列向量，每个元素的值都是0；
#### 2，取值范围中G、h的值
如果允许做空，则G为全0矩阵
如果不允许做空，则G为对角线为-1其他值为0的矩阵
为了保证权重向量都为正值，h为值=0的向量
#### 3，约束函数中A,b的值
约束函数中，权重向量加和为1；各类资产收益率的权重加和为目标收益率
A为第一行元素全为1、第二行元素为各类资产收益率的矩阵
b为形状为（2，1）的列向量，列向量的第一个元素值为1，第二个元素值为目标收益率

```python
# 导入所需要用的模块
import pandas as pd
import numpy as np
from cvxopt import matrix,solvers

# 自定义函数，用于数据检测
verify_num=1#数据检测次数初始化为1，每检测一次，次数+1，如果次数大于10，则终止检测
def verify_data(data):
        # 分割线
        print('-'*30)
        print '数据检测：'
        # 输出数据内容
        print(data)
        # 输出数据形状
        if isinstance(data,pd.DataFrame) or isinstance(data,pd.Series) or isinstance(data,np.ndarray):
            print '形状为:',data.shape
        if isinstance(data,list) or isinstance(data,dict):
            print 'len值为:',len(data)
        # 输出数据类型
        print '类型为:',type(data)
        
        # 如果次数大于1，终止检测
        global verify_num
        verify_num=verify_num+1
        if verify_num>10:
            raise ValueError('打印超过10次')
            
# 自定义函数
def describe(return_table,is_print=True):
    """以字典的形式，输出收益率矩阵的描述性统计量，包括：
    年化收益率
    年化标准差
    相关系数矩阵
    """
    
    # 导入gmean模块，计算几何平均值
    from scipy.stats.mstats import gmean
    # 初始化输出内容，空字典
    output={}
    
    """数据检测pd.DataFrame(dict(zip(return_table.columns,gmean(return_table+1.)**250-1.)),index=[0],columns=return_table.columns)：
indexID  000012.ZICN  000013.ZICN  000300.ZICN  000905.ZICN  399006.ZICN
0           0.045023     0.074658     0.068004     0.204963     0.404749
形状为: (1, 5)
类型为: <class 'pandas.core.frame.DataFrame'>"""
    """数据检测dict(zip(return_table.columns,gmean(return_table+1.)**250-1.))：
{'000300.ZICN': 0.068004214767582338, '000905.ZICN': 0.20496277012430908, ……'000012.ZICN': 0.045023247538068745}
len值为: 5
类型为: <type 'dict'>"""
    # zip()是Python的一个内建函数，它接受一系列可迭代的对象作为参数，将对象中对应的元素打包成一个个tuple（元组），然后返回由这些tuples组成的list（列表）。
    """数据检测zip(return_table.columns,gmean(return_table+1.)**250-1.)：
[('000012.ZICN', 0.045023247538068745), ('000013.ZICN', 0.074657634579693433), 
…… ('399006.ZICN', 0.4047493799342754)]
len值为: 5
类型为: <type 'list'>"""
    """数据检测gmean(return_table+1.)**250-1：
[ 0.04502325  0.07465763  0.06800421  0.20496277  0.40474938]
形状为: (5,)
类型为: <type 'numpy.ndarray'>"""
    # 先用gmean几何平均值算出来年化收益率；再用zip函数把return_table的列名和前面的年化收益率一一对应打包为元组，用元组组成列表；再用dict函数把上面的元组列表转换为字典；最后把字典转换为DataFrame格式的内容。整个过程参看上面的解释。index=[0]一定要保留
    output['annualized_return']=pd.DataFrame(dict(zip(return_table.columns,gmean(return_table+1.)**250-1.)),index=[0])
    
    """数据检测return_table.std()*np.sqrt(250)：
indexID
000012.ZICN    0.007540
000013.ZICN    0.005586
000300.ZICN    0.284462
000905.ZICN    0.325829
399006.ZICN    0.384782
dtype: float64
形状为: (5,)
类型为: <class 'pandas.core.series.Series'>"""
    # 获取日度收益率标准差，转换为年度，属于Series格式；转化为DataFrame格式；再转置
    output['annualized_volatility']=pd.DataFrame(return_table.std()*np.sqrt(250)).T

    """数据检测return_table.cov()：
indexID       000012.ZICN   000013.ZICN   000300.ZICN   000905.ZICN  \
indexID                                                               
000012.ZICN  2.273893e-07  3.732463e-08  1.852581e-07  3.543255e-07   
000013.ZICN  3.732463e-08  1.248103e-07  4.328221e-07  6.900750e-07   
形状为: (5, 5)
类型为: <class 'pandas.core.frame.DataFrame'>"""
    #从日度协方差到年度协方差；获取日度收益率协方差，乘以250得到年化协方差
    output['covariance_matrix']=return_table.cov()*250
    
    """数据检测return_table.corr()：
indexID      000012.ZICN  000013.ZICN  000300.ZICN  000905.ZICN  399006.ZICN
indexID                                                                     
000012.ZICN     1.000000     0.221557     0.021594     0.036058     0.011625
000013.ZICN     0.221557     1.000000     0.068097     0.094788     0.082960
……
形状为: (5, 5)
类型为: <class 'pandas.core.frame.DataFrame'>"""
    # 获取日度收益率相关系数
    output['coefficient_matrix']=return_table.corr()

    # 如果is_print的值为True，就打印出output字典的键和值
    if is_print:
        # 迭代output字典中的键值对
        for key,val in output.items():
            # 打印，{}是占位符，每个位置分别对应format括号中位置的内容；\n表示换行
            print "{}:\n{}\n".format(key,val)
            """打印样例：
            covariance_matrix:
indexID      000012.ZICN  000013.ZICN  000300.ZICN  000905.ZICN  399006.ZICN
indexID                                                                     
000012.ZICN     0.000057     0.000009     0.000046     0.000089     0.000034
000013.ZICN     0.000009     0.000031     0.000108     0.000173     0.000178
……"""
            
    return output

# 自定义函数
def get_efficient_frontier(return_table,allow_short=False,n_samples=25):
    """计算有效边界，返回'returns','risks','wegihts'"""
    from cvxopt import matrix,solvers
    # 获取资产类别
    assets=return_table.columns
    # 获取资产数量
    n_asset=len(assets)
    # 如果资产数量<2，报错
    if n_asset<2:
        raise ValueError('There must be at least 2 assets!')

    # 获取return_table的describe函数的返回值
    output=describe(return_table,is_print=False)
    """数据检测output：
{'covariance_matrix': 
indexID   000012.ZICN  000013.ZICN  000300.ZICN  000905.ZICN  399006.ZICN
indexID                                                                     
000012.ZICN     0.000057     0.000009     0.000046     0.000089     0.000034
000013.ZICN     0.000009     0.000031     0.000108     0.000173     0.000178
……, 
'annualized_return':  
    000012.ZICN  000013.ZICN  000300.ZICN  000905.ZICN  399006.ZICN
0     0.045023     0.074658     0.068004     0.204963     0.404749, 'annualized_volatility': 
indexID  000012.ZICN  000013.ZICN  000300.ZICN  000905.ZICN  399006.ZICN
0            0.00754     0.005586     0.284462     0.325829     0.384782, 'coefficient_matrix': 
indexID      000012.ZICN  000013.ZICN  000300.ZICN  000905.ZICN  399006.ZICN
indexID                                                                     
000012.ZICN     1.000000     0.221557     0.021594     0.036058     0.011625
000013.ZICN     0.221557     1.000000     0.068097     0.094788     0.082960
……}
len值为: 4
类型为: <type 'dict'>"""
    
    # 获取协方差矩阵，DataFrame格式；转换为np.array格式；最后转换为matrix格式
    covariance_matrix=matrix(output['covariance_matrix'].values)
    """数据检测output['covariance_matrix']：
indexID      000012.ZICN  000013.ZICN  000300.ZICN  000905.ZICN  399006.ZICN
indexID                                                                     
000012.ZICN     0.000057     0.000009     0.000046     0.000089     0.000034
000013.ZICN     0.000009     0.000031     0.000108     0.000173     0.000178
……
形状为: (5, 5)
类型为: <class 'pandas.core.frame.DataFrame'>"""
    """数据检测output['covariance_matrix'].values：
[[  5.68473354e-05   9.33115763e-06   4.63145157e-05   8.85813745e-05
    3.37267334e-05]
 [  9.33115763e-06   3.12025669e-05   1.08205521e-04   1.72518740e-04
    1.78310704e-04]
 ……]
形状为: (5, 5)
类型为: <type 'numpy.ndarray'>"""
    """数据检测matrix(output['covariance_matrix'].values)：
[ 5.68e-05  9.33e-06  4.63e-05  8.86e-05  3.37e-05]
[ 9.33e-06  3.12e-05  1.08e-04  1.73e-04  1.78e-04]
[ 4.63e-05  1.08e-04  8.09e-02  7.83e-02  7.11e-02]
[ 8.86e-05  1.73e-04  7.83e-02  1.06e-01  1.10e-01]
[ 3.37e-05  1.78e-04  7.11e-02  1.10e-01  1.48e-01]
类型为: <type 'cvxopt.base.matrix'>"""

    # 获取output['annualized_return']，DataFrame格式；切片之后变成Series格式；values之后变成array格式
    expected_return=output['annualized_return'].iloc[0,:].values
    """数据检测output['annualized_return'].iloc[0,:]：
000012.ZICN    0.045023
000013.ZICN    0.074658
000300.ZICN    0.068004
000905.ZICN    0.204963
399006.ZICN    0.404749
Name: 0, dtype: float64
形状为: (5,)
类型为: <class 'pandas.core.series.Series'>"""
    """数据检测output['annualized_return'].iloc[0,:].values：
[ 0.04502325  0.07465763  0.06800421  0.20496277  0.40474938]
形状为: (5,)
类型为: <type 'numpy.ndarray'>"""
    
    # 初始化风险、回报和权重，分别为空列表
    risks,returns,weights=[],[],[]
    
    # 应用二次规划模块solvers.qp；P,q是目标函数的协方差和一次项；G，h是不等式；A，b是等式
    # 迭代expected_return最小值和最大值之间的n_samples个样本；每个样本都是目标收益率
    for level_return in np.linspace(min(expected_return),max(expected_return),n_samples):
        # P是目标函数的协方差矩阵
        P=covariance_matrix#P为什么要乘以2啊？(X.T.dot(cov)).doc(X)就是总的方差，不用乘以2
        # q是目标函数的一次项；数组转化为matrix的列向量；后面做个练习，看看二维数组是怎么转化的
        q=matrix(np.zeros(n_asset))
        """数据检测np.zeros(n_asset)：
[ 0.  0.  0.  0.  0.]
形状为: (5,)
类型为: <type 'numpy.ndarray'>"""
        """数据检测q：
[ 0.00e+00]
[ 0.00e+00]
[ 0.00e+00]
[ 0.00e+00]
[ 0.00e+00]
类型为: <type 'cvxopt.base.matrix'>"""
        
        # 如果容许做空，则G是(n_asset,n_asset)形状数值为0的矩阵
        if allow_short:
            G=matrix(0.,(n_asset,n_asset))
        # 否则，如果不容许做空，则G是(n_asset,n_asset)形状的对角线数值为-1其他数值为0的矩阵
        else:
            G=matrix(np.diag(-1*np.ones(n_asset)))
        
        # 获取h值，形状为(n_asset,1)值为0的矩阵
        h=matrix(0.,(n_asset,1))
        
        
        # np.ones(n_asset)为元素值为1的数组，用于计算权重向量元素和等于1
        # expected_return为每个资产收益率组成的数组，用于计算权重向量乘积加和等于目标收益率
        # vstack沿着axis=0轴拼接；之后用matrix方法转化为矩阵
        A=matrix(np.vstack([np.ones(n_asset),expected_return]))
        """数据检测A：
[ 1.00e+00  1.00e+00  1.00e+00  1.00e+00  1.00e+00]
[ 4.50e-02  7.47e-02  6.80e-02  2.05e-01  4.05e-01]
类型为: <type 'cvxopt.base.matrix'>
"""
    
        # b为（2，1）的列向量，第一个元素值为1，第二个元素值为目标收益率；两个元素的列表可以通过matrix方法转化为（2，1）形状的列向量
        b=matrix([1.0,level_return])
        """数据检测b：
[ 1.00e+00]
[ 4.50e-02]
类型为: <type 'cvxopt.base.matrix'>"""
        
        # 是否显示解题过程
        solvers.options['show_progress']=False
        # 构造出了P,q,G,h,A,b，就可以用solvers.qp方法获取最小方差
        sol=solvers.qp(P,q,G,h,A,b)
        """数据检测sol：
{'status': 'optimal', 'dual slack': 2.186497302166682e-08, 'iterations': 6, 'relative gap': 0.0016565602346717302, 'dual objective': 2.840180270974816e-05, 'gap': 4.7049296961960596e-08, 'primal objective': 2.842366756152619e-05, 'primal slack': 9.157170806117054e-10, 's': <5x1 matrix, tc='d'>, 'primal infeasibility': 6.399762060694029e-09, 'dual infeasibility': 7.84603978692172e-09, 'y': <2x1 matrix, tc='d'>, 'x': <5x1 matrix, tc='d'>, 'z': <5x1 matrix, tc='d'>}
len值为: 14
类型为: <type 'dict'>"""
        
        # sol['primal objective']为目标收益率下的最小方差；把标准差加入到risks列表中
        risks.append(np.sqrt(sol['primal objective']))
        # 把目标收益率加入到returns列表中
        returns.append(level_return)
        # 把资产和权重用zip方法变成(单个资产,对应权重)元组的列表；用dict方法把资产和权重转化为字典；把字典加入到weights列表
        weights.append(dict(zip(assets,list(sol['x'].T))))
        """数据检测sol['x']：
[ 1.00e+00]
[-1.50e-10]
[ 5.12e-09]
[-1.95e-09]
[ 5.51e-10]
类型为: <type 'cvxopt.base.matrix'>"""
        """数据检测sol['x'].T：
[ 1.00e+00 -1.50e-10  5.12e-09 -1.95e-09  5.51e-10]
类型为: <type 'cvxopt.base.matrix'>"""
        
    # 迭代结束后，建立output字典，键值对分别为收益率、风险和权重
    output={'returns':returns,
           'risks':risks,
           'weights':weights}
    """数据检测output：
{'returns': [0.045023247538068745, 0.052364597178807655, 0.059705946819546565, 
……0.3974080302935365, 0.4047493799342754], 
'weights': [{'000300.ZICN': 5.115357855584461e-09, '000905.ZICN': -1.946343551896132e-09, '399006.ZICN': 5.50919147696566e-10, '000013.ZICN': -1.4977963328537784e-10, '000012.ZICN': 0.9999999964298462}, 
    {'000300.ZICN': 0.0006194710901677276, '000905.ZICN': 1.3937363356247555e-05, '399006.ZICN': 8.563182291128989e-06, '000013.ZICN': 0.2470712166222291, '000012.ZICN': 0.7522868117419558}, ……], 
'risks': [0.0053313851447373594, 0.0043393048365158041, 0.003665097414747384, 
……0.26603820850802012, 0.27208209558346064]}
len值为: 3
类型为: <type 'dict'>"""
    
    # 把字典格式转化为DataFrame格式
    output=pd.DataFrame(output)
    """数据检测output：
     returns     risks                                            weights
0   0.045023  0.005331  {u'000300.ZICN': 5.11535785558e-09, u'000905.Z...
1   0.052365  0.004339  {u'000300.ZICN': 0.000619471090168, u'000905.Z...
2   0.059706  0.003665  {u'000300.ZICN': 0.000388864228719, u'000905.Z...
形状为: (50, 3)
类型为: <class 'pandas.core.frame.DataFrame'>    """
    
    return output


def draw_efficient_frontier(efficient_frontier_output):
    import seaborn
    import matplotlib.pyplot as plt
    
    """数据检测efficient_frontier_output：
     returns     risks                                            weights
0   0.045023  0.005331  {u'000300.ZICN': 5.11535785558e-09, u'000905.Z...
1   0.052365  0.004339  {u'000300.ZICN': 0.000619471090168, u'000905.Z...
2   0.059706  0.003665  {u'000300.ZICN': 0.000388864228719, u'000905.Z...
形状为: (50, 3)
类型为: <class 'pandas.core.frame.DataFrame'>"""

    # 设置画布
    fig=plt.figure(figsize=(7,4))
    # 在画布上增加子图
    ax=fig.add_subplot(111)
    # 在子图上画图；除了array数据，Serie数据也可以用于画图
    ax.plot(efficient_frontier_output['risks'],efficient_frontier_output['returns'])
    """数据检测efficient_frontier_output['risks']：
0     0.005331
1     0.004339
2     0.003665
Name: risks, dtype: float64
形状为: (50,)
类型为: <class 'pandas.core.series.Series'>"""

    # 设置标题
    ax.set_title('Efficient Frontier',fontsize=14)
    # 为X轴设置标签
    ax.set_xlabel('Standard Deviation',fontsize=12)
    # 为Y轴设置标签
    ax.set_ylabel('Expected Return',fontsize=12)
    # 设置刻度值的字体的大小
    ax.tick_params(labelsize=12)
    plt.show()
    
# 获取最小方差的组合    
def get_minimum_variance_portfolio(return_table,allow_short=False,show_details=True):
    import numpy as np
    from cvxopt import matrix,solvers
    
    # 获取各类资产
    assets=return_table.columns
    # 获取资产数量
    n_asset=len(assets)
    # 如果资产数量小于2，只有一种资产，则资产的权重设置为1
    if n_asset<2:
        weights=np.array([1.0])
        weights_dict={assets[0]:1.0}
    # 否则，如果资产数量大于或者等于2，通过solvers.qp函数计算最优权重
    else:
        # 通过describe函数获取协方差矩阵、年化收益和相关系数矩阵
        output=describe(return_table,is_print=False)
        # 用matrix方法把array数组转化为矩阵
        covariance_matrix=matrix(output['covariance_matrix'].values)
        # 获取年化收益，array数组格式
        expected_return=output['annualized_return'].iloc[0,:].values
        
        # P为协方差矩阵
        P=covariance_matrix
        # q是值为0的列向量
        q=matrix(np.zeros(n_asset))
        
        # 如果允许做空，G为元素全是0的矩阵
        if allow_short:
            G=matrix(0.0,(n_asset,n_asset))
        # 否则，如果不允许做空，G为对角线是-1其他元素全是0的矩阵
        else:
            G=matrix(np.diag(-1*np.ones(n_asset)))
        
        # h是列向量，元素全是0
        h=matrix(0.0,(n_asset,1))
        
        # np.ones(n_asset)为元素全是1的array数组；使用matrix方法后变成了列向量；转置后变成了行向量；权重向量的加和为1
        A=matrix(np.ones(n_asset)).T
        """数据检测A：
[ 1.00e+00  1.00e+00  1.00e+00  1.00e+00  1.00e+00]
类型为: <type 'cvxopt.base.matrix'>"""
        
        # 权重向量的加和为1；无目标收益率的限制
        b=matrix([1.0])
        # 设置选项，不展示求解过程
        solvers.options['show_progress']=False
        # 确定P,q,G,h,A,b后，通过solvers.qp方法求解
        sol=solvers.qp(P,q,G,h,A,b)
        """数据检测sol：
{'status': 'optimal', 'dual slack': 9.986070814625971e-10, 'iterations': 8, 'relative gap': 0.003088348526373338, 'dual objective': 1.2148774765538435e-05, 'gap': 3.751965064439222e-08, 'primal objective': 1.2186294416182828e-05, 'primal slack': 7.411627205758946e-05, 's': <5x1 matrix, tc='d'>, 'primal infeasibility': 1.2412677424910066e-16, 'dual infeasibility': 6.65219003640825e-20, 'y': <1x1 matrix, tc='d'>, 'x': <5x1 matrix, tc='d'>, 'z': <5x1 matrix, tc='d'>}
len值为: 14
类型为: <type 'dict'>"""
        
        weights=np.array(sol['x'].T)[0]
        """数据检测sol['x']：
[ 3.16e-01]
[ 6.84e-01]
[ 1.40e-04]
[ 7.41e-05]
[ 9.09e-05]
类型为: <type 'cvxopt.base.matrix'>"""
        """数据检测np.array(sol['x'].T)：
[[  3.15530952e-01   6.84164304e-01   1.39698074e-04   7.41162721e-05
    9.09298343e-05]]
形状为: (1, 5)
类型为: <type 'numpy.ndarray'>"""
        """数据检测weights：
[  3.15530952e-01   6.84164304e-01   1.39698074e-04   7.41162721e-05
   9.09298343e-05]
形状为: (5,)
类型为: <type 'numpy.ndarray'>"""
        
        weights1=list(sol['x'].T)
        """数据检测weights1：
[0.31553095216683397, 0.684164303652688, 0.00013969807415992067, 7.41162720575894e-05, 9.092983426059725e-05]
len值为: 5
类型为: <type 'list'>"""
        
        # zip函数把（单一资产，对应权重）打包成元组形式以组成列表；zip函数把元组转化为字典
        weights_dict=dict(zip(assets,weights))
    
    # 对权重数组和资产年化收益率数组使用dot函数，得到组合收益率
    r=np.dot(weights,output['annualized_return'].iloc[0,:].values)
    # 对权重数组和协方差矩阵使用dot函数，得到值和weights.T使用dot函数，得到方差；求平方根得到标准差
    v=np.sqrt(np.dot(np.dot(weights,covariance_matrix),weights.T))
    
    if show_details:
        print """
        Minimum Variance Portfolio:
        Short Allowed:{}
        Portfolio Return:{}
        Portfolio Volatility:{}
        Portfolio Weights:{}
        """.format(
            allow_short,r,v,"\n\t{}".format(
                "\n\t".join("{}:{:.1%}".format(k,v) for k,v in weights_dict.items()))).strip()
        
        # join函数的使用，用"\n\t"连接"{}:{:.1%}"的内容；0.1%表示仅保留1位小数
        abc="\n\t".join("{}:{:.1%}".format(k,v) for k,v in weights_dict.items())
        """数据检测abc：
000300.ZICN:0.0%
	000905.ZICN:0.0%
	399006.ZICN:0.0%
	000013.ZICN:68.4%
	000012.ZICN:31.6%
类型为: <type 'str'>"""
        # 把abc放进{}占位符中；由下面的数据监测可知，的确体现了\n\t的效果
        cde="\n\t{}".format(abc)
        """数据检测cde：

	000300.ZICN:0.0%
	000905.ZICN:0.0%
	399006.ZICN:0.0%
	000013.ZICN:68.4%
	000012.ZICN:31.6%
类型为: <type 'str'>"""
                
    return weights_dict
    
# 获取效用最大化的组合；效用函数中含有二次项和一次项，因此，P和q的值要重新定义
def get_maximum_utility_portfolio(return_table,risk_aversion=3.,
    allow_short=False,show_details=True):
    # 获取所有的资产种类 
    assets=return_table.columns
    # 获取资产的数量
    n_asset=len(assets)
    # 如果资产数量小于2，则该资产的权重设为1
    if n_asset<2:
        weights=np.array([1.])
        weights_dict={assets[0]:1.}
    # 否则，如果资产数量大于等于2，则应用solvers.qp二次规划求解
    else:
        # 通过describe函数获取return_table的统计数据
        output=describe(return_table,is_print=False)
        # 在describe函数中，协方差矩阵已经乘以了250；这里，不再乘以250；values获取了array数组；通过matrix方法得到矩阵
        covariance_matrix=matrix(output['covariance_matrix'].values)
        # 通过iloc获取Series；通过values获取array数组
        expected_return=output['annualized_return'].iloc[0,:].values
        
        # 当风险厌恶系数小于0.01时，0.5*A*方差可以忽略，效用最大化的值就是各资产预期回报的最大值
        if abs(risk_aversion)<0.01:
            # 最大回报就是各资产预期回报的最大值
            max_ret=max(expected_return)
            # 获取权重；如果该资产回报等于最大回报，则权重为1；否则，如果该资产回报不等于最大回报，则权重为0
            weights=np.array([1. if expected_return[i]==max_ret else 0. for i in range(n_assets)])
            # 建立权重字典，键为资产，值为对应的权重
            weights_dict={asset:weights[i] for i,asset in enumerate(assets)}
            
        # 否则，如果风险厌恶系数大于0.01，则按照正常方法计算；P值为0.5*A*方差；q值为负的各资产预期收益的列向量
        else:
            # P值为0.5*A*方差
            P=0.5*risk_aversion*covariance_matrix
            # q值为负的各资产预期收益的列向量；array和matrix的转化：如果array是多维数组，(n,m)形状，则按照(n,m)形状转化为matrix；如果array是一维数组，(m,)形状，则按照(m,1)形状转化为matrix
            q=matrix(-expected_return.T)
            # 对于一维数组，(m,)形状的转置依然是(m,)形状
            """数据检测expected_return：
[ 0.04502325  0.07465763  0.06800421  0.20496277  0.40474938]
形状为: (5,)
类型为: <type 'numpy.ndarray'>
------------------------------
数据检测expected_return.T：
[ 0.04502325  0.07465763  0.06800421  0.20496277  0.40474938]
形状为: (5,)
类型为: <type 'numpy.ndarray'>
------------------------------
数据检测q：
[ 4.50e-02]
[ 7.47e-02]
[ 6.80e-02]
[ 2.05e-01]
[ 4.05e-01]
类型为: <type 'cvxopt.base.matrix'>"""
            
            
            # 如果允许做空，G为元素全是0的矩阵
            if allow_short:
                G=matrix(0.,(n_asset,n_asset))
            # 否则，如果不允许做空，G为对角线是-1其他元素全是0的矩阵
            else:
                G=matrix(np.diag(-1*np.ones(n_asset)))
                
            # 获取h值，形状为(n_asset,1)值为0的矩阵
            h=matrix(0.,(n_asset,1))
            # np.ones(n_asset)为元素全是1的array数组；使用matrix方法后变成了列向量；转置后变成了行向量；权重向量的加和为1
            A=matrix(np.ones(n_asset)).T
            # 权重向量的加和为1
            b=matrix([1.0])
            # slovers方法的选项，不显示计算过程
            solvers.options['show_progress']=False
            # 获取P,q,G,h,A,b后，通过solvers.qp获取解决方案
            sol=solvers.qp(P,q,G,h,A,b)
            # 权重向量
            weights=np.array(sol['x'].T)[0]
            # 把资产类别和权重向量一一对应，写入weights_dict字典
            weights_dict=dict(zip(assets,weights))
            
    # 获取组合回报————权重向量与年化收益率向量的点积
    r=np.dot(weights,output['annualized_return'].iloc[0,:].values)
    # 获取风险————权重向量与协方差矩阵点积，再与权重向量的转置点积，最后开方
    v=np.sqrt(np.dot(np.dot(weights,covariance_matrix),weights.T))
    
    
    # 如果要显示细节，则打印出来；print可以结合""""""使用，保留其中的格式；strip()去掉头尾的\n\t等字符
    if show_details:
        # {}是占位符，分别对应.format()中以,号隔开的字符；Portfolio Weights:{}占位符对应"\n\t{}".format()的内容；"{}:{:0.1%}"对应其后.format(k,v)中的内容，0.1%表示只取一位小数
        print """
        Maximum Utility Portfolio:
        Risk Aversion:{}
        Short Allowed:{}
        Portfolio Return:{}
        Portfolio Volatility:{}
        Portfolio Weights:{}
        """.format(risk_aversion,allow_short,r,v,
                   "\n\t{}".format(
                            "\n\t".join("{}:{:0.1%}".format(k,v) for k,v in weights_dict.items()))).strip()
        
    return weights_dict
    
# 自定义函数，获取产生最大夏普比率的组合
def get_maximum_sharpe_portfolio(return_table,riskfree_rate=0.,allow_short=False,show_details=True):
    assets=return_table.columns
    n_asset=len(assets)
    if n_asset<2:
        output=describe(return_table,is_print=False)
        # iat可以用iloc取代
        r=output['annualized_return'].iat[0,0]
        v=output['annualized_volatility'].iat[0,0]
        weights_dict={assets[0]:1.}
    else:
        # 获取有效边界，回报最小值与最大值之间的样本数为100
        efs=get_efficient_frontier(return_table,allow_short=allow_short,n_samples=100)
        """数据检测efs：
     returns     risks                                            weights
0   0.045023  0.005186  {u'000300.ZICN': 0.00502839510695, u'000905.ZI...
1   0.048657  0.004694  {u'000300.ZICN': 0.00440441690257, u'000905.ZI...
2   0.052290  0.004259  {u'000300.ZICN': 0.0037804386982, u'000905.ZIC...
[100 rows x 3 columns]
类型为: <class 'pandas.core.frame.DataFrame'>"""
        
        # max函数的使用，找出key函数值最大时对应的range(100)中的值，这个值正好是索引值；key中的lambda函数是夏普比率计算公式
        i_star=max(range(100),key=lambda x:(efs.at[x,'returns']-riskfree_rate)/efs.at[x,'risks'])
        
        # 通过行列索引获取回报
        r=efs.at[i_star,'returns']
        # 通过行列索引获取风险
        v=efs.at[i_star,'risks']
        # 通过行列索引获取资产及其对应的权重
        weights_dict=efs.at[i_star,'weights']
        
    # 计算夏普比率
    s=(r-riskfree_rate)/v
    
    # 如果需要显示细节，则把""""""中第一个.format之前的{}占位符填满，每个占位符分别对应.format()括号中的变量；Portfolio Weights:{}占位符对应的内容是下一个占位符，该占位符内容由"\n\t{}".format()决定；"\n\t{}".format()占位符的内容由后面的join函数连接的字符串决定；三级占位符，逐层理解
    if show_details:
        print """
        Maximum Sharpe Portfolio:
        Riskfree Rate:{}
        Short Allowed:{}
        Portfolio Return:{}
        Portfolio Volatility:{}
        Portfolio Sharpe:{}
        Portfolio Weights:{}
        """.format(riskfree_rate,allow_short,r,v,s,
                   "\n\t{}".format(
                            "\n\t".join("{}:{:0.1%}".format(k,v) for k,v in weights_dict.items()))).strip()
    
    return weights_dict
    
    
# 初始变量设置
start='20130101'
end='20160630'
indices=['000300.ZICN','000905.ZICN','399006.ZICN','000012.ZICN','000013.ZICN']

# 获取股票指数日度收盘指数数字
df=DataAPI.MktIdxdGet(indexID=indices,
                     beginDate=start,
                     endDate=end,
                     field="tradeDate,indexID,closeIndex")
"""数据检测df：
       tradeDate      indexID  closeIndex
0     2013-01-04  000012.ZICN     135.840
1     2013-01-07  000012.ZICN     135.875
[4235 rows x 3 columns]
类型为: <class 'pandas.core.frame.DataFrame'>"""

# 在这里使用了书中没有提到的函数，用以获取指数对应的简称
name_indices=DataAPI.IdxGet(secID=indices,
                            field=u"secID,secShortName",
                            pandas="1")
print(name_indices)

# 通过pivot把长数据压扁为宽数据，方便查看
df=df.pivot(index='tradeDate',
           columns='indexID',
           values='closeIndex')
"""数据检测df：
indexID     000012.ZICN  000013.ZICN  000300.ZICN  000905.ZICN  399006.ZICN
tradeDate                                                                  
2013-01-04     135.8400     159.6470    2524.4100    3258.2510      705.342
2013-01-07     135.8750     159.7240    2535.9900    3288.3690      718.114
2013-01-08     135.8790     159.7620    2525.3300    3327.0400      733.138
[847 rows x 5 columns]
类型为: <class 'pandas.core.frame.DataFrame'>"""

# 通过pct_change()函数实现下一行用迭代才能实现的功能
df=df.pct_change().dropna()

# # 迭代indices中的每个元素，即指数代码；计算对应日期的收益率
# for index in indices:
#     # 通过shift函数实现了对Series中索引的移动
#     df[index]=df[index]/df[index].shift()-1

return_table=df.dropna()
output=describe(return_table,is_print=False)
output
```
