1.1 为什么要检测有效因子的独立性
-----------------
>假设，市场中存在n个参量（维度），互相独立，并可以完备地（非常好地）描述市场的属性。如果我们能够找出这些参量，理论上就可以通过历史数据非常好的预测市场未来的趋势。这与在解析几何空间中找到一套坐标系，或者是在量子力学希尔伯特空间中寻找一套完备的基矢的过程类似。但是，一般认为市场中包含大量的随机因素，市场趋势具有一定随机性的行为，这表明我们很难通过推导的方式去计算出一套完备的基矢去描述市场。我们能做的是尽可能的去找到一些可以描述市场，同时互相之间独立（包含不同信息）的参量，利用这样的‘准完备基矢’逼近地描述市场，以期获得不错的市场预测。
>
>在上一部分中，经过筛选得到了67个有效因子，其有效性表明这些因子可以从不同方面去描述市场。但是这些因子并不是互相独立的，多个因子之间存在或强或弱的相似性，即存在很多的重复。为了获得对市场描述的不同维度，以及节省计算资源，我们需要检测这些因子间的相关性，最终选取出低相关性，即互相独立的因子，作为我们的‘准完备基矢’。

1.2 思路
------------------

>1.计算所有有效因子的相关系数矩阵；
>2.筛选，对于具有强相关性的因子对，依据在上一部分中计算的两者的收益情况，舍弃低收益因子，保留高收益因子；

1.3 具体操作方法
------------------
>1.为了提高计算速度，首先读取67个有效因子2008年初到2013年末的历史数据，组成一组DataFrame数据，保存在存储空间中。
>2.由于筛选出的67个有效因子中，部分因子数值存在频繁剧烈地波动，这会在因子择股策略中造成很高的换手率，提高了交易成本，因此首先对67个因子进行筛选，过滤掉这种会带来高换手率的因子。
>3.计算每个因子在每交易日股票排序，这是为了‘标准化’因子的数值。
>4.计算HS300成分股每只股票各有效因子排序的相关性系数矩阵，并对所有股票求平均，得到各因子平均相关性系数矩阵。
>5.分级地对不同相关强度的因子对进行筛选处理，保留高收益因子，删除低收益因子，经过几重循环计算最终可以的到一定数量的互相独立的有效因子。

>1.4.1 为了提高计算速度，首先读取67个有效因子2008年初到2013年末的历史数据，组成一组DataFrame数据，并保存在存储空间中.

```python
import pandas as pd
from pandas import Series, DataFrame

def read_factor(daytime, stock):  # 读取因子数据的函数，每次只能一个交易日数个股票的数据，注意读取的数据中股票secID码的顺序会重新排序
    f = DataAPI.MktStockFactorsOneDayGet(daytime,secID=stock,ticker=u"",field=u"",pandas="1").set_index(['secID', 'tradeDate']).drop('ticker', axis=1)
    return f

############# 读取交易日数据 ##############
# beginDate=u"20080101",endDate=u"20131231"
date=DataAPI.TradeCalGet(exchangeCD=u"XSHG",beginDate=u"20080101",endDate=u"20131231",field=u"calendarDate,isOpen,isMonthEnd",pandas="1")
tradeDate = date[date['isOpen']==1]['calendarDate']
tradeDate.index = tradeDate

############# 读取因子历史数据 ##############
effective_factor = pd.read_csv("Multi-factor Model_2/effective_factor.csv", index_col='factor')
effective_factor_list = effective_factor.index.tolist()
effective_factor_data = DataFrame()
for day in tradeDate:
    stocks = DataAPI.IdxConsGet(secID=u"",ticker=u"000300",intoDate=day,isNew=u"",field=u"",pandas="1")['consID']
    data = read_factor(day, stocks)[effective_factor_list]
    effective_factor_data = pd.concat([effective_factor_data, data])
    if (date[date['calendarDate']==day]['isMonthEnd']==1).all():
        print day
        
effective_factor_data.to_csv('Multi-factor Model_2/effective_factor_data_08_13.csv')  # 保存
```

>1.4.2 由于筛选出的67个有效因子中，部分因子数值存在频繁地剧烈地波动，这会在因子择股策略中造成很高的换手率，提高了交易成本，因此首先对67个因子进行筛选，过滤掉这种会带来高换手率的因子。
>
>方法为：计算各因子自2008年初到2013年末对所有股票历史数据的标准差‘std_raw’，然后计算经过20天平均的数据的标准差‘std_mean’，求两者比值，比值大于一定值，认为该因子数据具有很强的波动性，舍弃该因子。

```python
import pandas as pd
from pandas import Series, DataFrame
effective_factor_data = pd.read_csv('Multi-factor Model_2/effective_factor_data_08_13.csv', index_col=['tradeDate', 'secID']).unstack()
effective_factor = pd.read_csv("Multi-factor Model_2/effective_factor.csv", index_col='factor')
factor_list = effective_factor.index.tolist()
ratio = Series()
for factor in factor_list:
    std_raw = effective_factor_data[factor].std()
    std_mean = pd.rolling_mean(effective_factor_data[factor], 20).std()
    ratio_each_factor = (std_raw.mean())/(std_mean.mean())
    ratio[factor] = ratio_each_factor
ratio.order().plot()
print ratio.order()

effective_factor_list = ratio.order()[:-28].index.tolist()
#effective_factor_list.append('TA2EV')
effective_factor_data = effective_factor_data[effective_factor_list]  #将因子数据置换为‘低换手率’因子数据
effective_factor_data.to_csv('Multi-factor Model_2/effective_factor_data_2_08_13.csv)
```

>1.4.3 计算每个因子在每交易日股票排序，这是为了‘标准化’因子的数值。(我另外在不做这种标准化的情况下计算了因子间的相关性矩阵，发现差别很小。)

```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
effective_factor_data = pd.read_csv('Multi-factor Model_2/effective_factor_data_2_08_13.csv', header=[0, 1], index_col=0)
effective_factor_list = effective_factor_data.columns.levels[0].tolist()
effective_factor_rank_data = DataFrame()
for factor in effective_factor_list:
    rank_for_each_factor = effective_factor_data[factor].rank(axis=1)
    rank_for_each_factor.columns = [[factor]*len(rank_for_each_factor.columns), rank_for_each_factor.columns]
    effective_factor_rank_data = pd.concat([effective_factor_rank_data, rank_for_each_factor], axis=1)
```

>1.4.4 计算HS300成分股每只股票各有效因子排序的相关性系数矩阵，并对所有股票矩阵求平均，得到各因子相关性系数矩阵。

```python
data_for_corr = effective_factor_rank_data.swaplevel(1, 0, axis=1)
stocks = data_for_corr.columns.levels[0].tolist()
correlation = DataFrame()
for sto in stocks:  # 计算每只股票的各因子相关系数矩阵。
    corr_sto = data_for_corr[sto].corr()
    corr_sto.index = [[sto]*len(corr_sto.index), corr_sto.index]
    correlation = pd.concat([correlation, corr_sto])
correlation_mean = correlation.mean(axis=0, level=1).sort_index(axis=1).sort_index(axis=0)   # 计算平均相关系数矩阵。
#correlation_std = correlation.std(axis=0, level=1).sort_index(axis=1).sort_index(axis=0)
```

>1.4.5 分级地对不同相关强度的因子对进行筛选处理，保留高收益因子，删除低收益因子，经过几重循环计算最终可以的到一定数量的互相独立的有效因子。
>
>相关系数分级从0.95到0.35，间隔为0.5。分级的计算是为了防止这种因子筛选可能的路径依赖问题，最终得到的因子间相关性低于0.35。

```python
effective_factor_info = pd.read_csv("Multi-factor Model_2/effective_factor.csv", index_col='factor')
c = correlation_mean.copy()
#drop_list = []
for f_1 in correlation_mean:   # 首先丢弃收益低于一定值的因子。
    if f_1 in c:
        if effective_factor_info['hedged earn'][f_1] < 0.47:
            c = c.drop(f_1, axis=1).drop(f_1, axis=0)
#            drop_list.append(f_1)

reference = np.arange(0.95, 0.30, -0.05)  # 相关系数分级
for ref in reference:    # 分级地筛选因子，降低可能地路径依赖问题。
    for f_1 in correlation_mean:
        if f_1 in c:
            corr_f = c[f_1][(c[f_1].abs()>ref)&(c[f_1]!=1)]  # the bigger the constant is, the more factors left.
            for f_2 in corr_f.index:
                drop_f =  effective_factor_info['hedged earn'][[f_1, f_2]].abs().idxmin()
                c = c.drop(drop_f, axis=1).drop(drop_f, axis=0)
#                drop_list.append(drop_f)
                if drop_f == f_1:
                    break
independent_factor_info = effective_factor_info.ix[c.index].sort_index(by='mean')
#independent_factor_info.to_csv("Multi-factor Model_2/independent_factor_info.csv")
#c.to_csv("Multi-factor Model_2/independent_factor_corr.csv")
#print len(drop_list)
print 'The independent factors:'
print c.index.tolist()
```
