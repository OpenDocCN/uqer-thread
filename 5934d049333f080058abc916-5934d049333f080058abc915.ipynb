```python
import sklearn as sk
import pandas as pd
import math
from CAL.PyCAL import *
cal = Calendar('China.SSE')
from datetime import datetime
```

```python
def getStkList(pday):
    stkList = DataAPI.MktEqudGet(tradeDate=pday,secID=u"",ticker=u"",beginDate=u"",endDate=u"",isOpen="",field=u"secID",pandas="1")['secID'].tolist()
    tDayFactorTable = DataAPI.MktStockFactorsOneDayGet(tradeDate=pday,secID=stkList,ticker=u"",field=u"secID,revs20,DAVOL10,DAVOL20",pandas="1")
    AllRevDown = tDayFactorTable.mean()[0]<0.97
    AllVolumeUp = tDayFactorTable.mean()[1]/tDayFactorTable.mean()[2]-1>0.05
    sortedList = tDayFactorTable.sort('REVS20')['secID'].tolist()#[:len()]
    bottom10List = sortedList[:len(sortedList)/10]
    return AllRevDown,AllVolumeUp,bottom10List
```
