# “漂亮50”粗筛

参考股吧文章《A股分两种：“漂亮50”和“要命3000”！三大指标筛选“漂亮50”(附名单)》做了一个类似的筛选。
原文链接 http://guba.eastmoney.com/news,cjpl,634873730.html
所谓“漂亮50”，是来自于美国的概念，中国投资者很早也提出来要筛选出类似公司，但是一直没有核心指标。
证券时报记者以三个指标筛选出A股的“漂亮50”，这三个指标分别是净利润增长率长大于15%，连续3年净资产收益率大于15%，市盈率低于35.

```python
import time

#生成A股列表
#获取运行当日日期
current_date = time.strftime("%Y%m%d",time.localtime(time.time()))
print "Today is " + str(current_date)

#净利润增长率筛选 大于15%
data = DataAPI.FdmtIndiGrowthGet(ticker=u"",secID="",endDate="",beginDate="",beginYear=u"2016",endYear=u"2017",reportType=u"A",field=u"ticker,niYOY",pandas="1")
#print data
data = data[data.niYOY > 15]
#print data
data = data[data.ticker < "900000"]
#print data
xp = data['ticker'].values.tolist()
print xp


#PE筛选 低于35
data_pe = DataAPI.MktStockFactorsOneDayGet(tradeDate=u"20170502",secID=u"",ticker=xp,field=u"ticker,tradeDate,pe",pandas="1")
data_pe = data_pe[data_pe.PE < 35]
xp = data_pe['ticker'].values.tolist()
print xp

#连续3年净资产收益率大于15%
win = []
for x in xp:
    #print x
    data_roe = DataAPI.FdmtIndiRtnGet(ticker=x,secID="",endDate="",beginDate="",beginYear=u"2014",endYear=u"2017",reportType=u"A",field=u"ROEW",pandas="1")
    if len(data_roe) == 3:
        roe_2014 = data_roe.ix[0,"ROEW"]
        roe_2015 = data_roe.ix[1,"ROEW"]
        roe_2016 = data_roe.ix[2,"ROEW"]
        if roe_2014 > 15 and roe_2015 > 15 and roe_2016 > 15:
            win.append(x)
            print x
            
total_num = len(win)
print "The total number is " + str(total_num)
```

共筛选出59只股票，与文章中的56只大多一样。差异可能是由于所选时间不容造成的。
需要注意的是DataAPI.FdmtIndiRtnGet函数中资产收益率需要用“ROEW”(净资产收益率(加权平均,%) , 直接取公告披露值)而不是“ROE”(净资产收益率(摊薄,%) , ROE(摊薄)=归属于母公司的净利润/期末归属于母公司的所有者权益*100%)这样才能和原文中的筛选结果类似。
