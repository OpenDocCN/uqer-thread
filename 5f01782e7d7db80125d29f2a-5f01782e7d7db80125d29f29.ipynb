```python
import numpy as  np
import scipy as sp
import pandas as pd
import datetime
from CAL.PyCAL import *
import os

# 股票池文件存放目录，如果目录不存在，程序自动新建一个
raw_data_dir = "./FR_factor"
if not os.path.exists(raw_data_dir):
    os.mkdir(raw_data_dir)
```

```python
#定义几个需要的函数：

#定义时间转换函数  将日期由2018-03-12 转化为20180312
def time_change(x):
    y = datetime.datetime.strptime(x,'%Y-%m-%d')
    z = y.strftime('%Y%m%d')
    return z
#time_change('2018-09-23')

# 定义获取所有交易日历，月末一天，月初一天的list函数
def get_trade_list(start_date,end_date):
    cal = Calendar('China.SSE')
    all_day = DataAPI.TradeCalGet(exchangeCD=u"XSHG",beginDate=start_date,endDate=end_date,\
                                  isOpen=u"1",field=u"calendarDate,isOpen,isMonthEnd",pandas="1")
    
    #获取所有的交易日历 trade_list (list型)
    trade_list = all_day[all_day['isOpen']==1]['calendarDate'].tolist()
    trade_list = [time_change(x) for x in trade_list]
    
    #获取所有的交易月末最后一天 month_end (list型)
    month_end = all_day[all_day['isMonthEnd']==1]['calendarDate'].tolist()
    month_end = [time_change(x) for x in month_end]
    
    #获取所有的交易月初第一天  month_start (list型)
    month_start = [cal.advanceDate(x,'1B') for x in month_end]
    month_start = [x.strftime('%Y%m%d') for x in month_start]
    
    return trade_list,month_end,month_start

#get_trade_list('20071220','20180601')

#定义函数去掉ST股票
def st_remove(source_universe,st_date =None):
    st_date = st_date if st_date is not None else datetime.datetime.now().strftime('%Y%m%d')
    st_secid = DataAPI.SecSTGet(beginDate=st_date,endDate=st_date,secID=source_universe,field=u"secID",pandas="1")
    return [x for x in source_universe if x not in list(st_secid['secID'])]

#定义函数去掉新股   ticker为6位数值，没有后缀  注意日期之间的比较
def new_remove(ticker,trade_date =None,day=20):
    
    trade_date = trade_date if trade_date is not None else datetime.datetime.now().strftime('%Y%m%d')
    
    cal = Calendar('China.SSE')
    sta_date = cal.advanceDate(trade_date,str(str(day*(-1))+'B')).strftime('%Y-%m-%d')
    
    new_ticker = DataAPI.SecIDGet(assetClass=u"e",ticker = ticker,field=u"ticker,listDate",pandas="1")
    new_ticker=new_ticker[new_ticker['listDate']>sta_date]['ticker'].tolist()
    
    return [x for x in ticker if x not in new_ticker]


#将ticker转化为函数内部编码
def ticker2secID(ticker):
    
    #DateFrame生成字典的方式，标准
    all_a = DataAPI.EquGet(equTypeCD=u"A",field=u"secID,ticker",listStatusCD='L,S,DE,UN',pandas="1")
    all_a_dict = dict(all_a.set_index(all_a['ticker'])['secID'])
    
    if isinstance(ticker,list):
        secID_list = []
        for i in ticker:
            if i in all_a_dict.keys():
                secID_list.append(all_a_dict[i])
            else:
                print(str(i)+'不在字典中')
            
    else:
        print('输入的不是list形式')
    return secID_list

#定义函数，获取每月月初的股票代码，作为以后进行研究的股票代码基础
def get_stock_pool(date,n):  #date  每一个月月初的日期  n 界定新股时候的参数
    #该函数的目的为获得每个月第一个交易日的股票代码，加上那个日期，构成DF
    
    univ = DynamicUniverse('A')
    code = univ.preview(date,skip_halted='False') #返回list  
    code = st_remove(code,st_date =date)  #去掉ST
    code = [x[0:6] for x in code]  # 代码去掉后缀
    code = new_remove(code,trade_date =date,day=n)  #去掉新股
    code = ticker2secID(code) #加上后缀
    code_df = pd.DataFrame(dict({'code':code}))  # list 转化为字典  再转化为DF
    code_df['date']=date
    code_df = code_df[['date','code']]
    return code_df  # 为df形式
```

```python
#获得整个研究区间每个月月初所有的股票
start_date = '20071220'
end_date = '20180601'
trade_list,month_end,month_start = get_trade_list(start_date,end_date)
n=180
all_stock = pd.DataFrame()
for date in month_start:
    stock = get_stock_pool(date,n)
    all_stock = pd.concat([all_stock,stock],axis = 0)
    #if i>'20150602':
    #   break
all_stock.head()
```

```python
all_stock.to_csv('FR_factor/stock_pool.csv', index=False)
```

```python
#获取整个研究期间所有交易日的股票代码的日换手率以及涨跌幅
all_code = sorted(list(set(list(all_stock['code']))))
mkt_info_frame = DataAPI.MktEqudGet(secID=all_code,beginDate=u"20071220",endDate=u"20180601",isOpen=1,\
                                    field=u"tradeDate,secID,chgPct,turnoverRate",pandas="1")
mkt_info_frame['tradeDate']=map(time_change,mkt_info_frame['tradeDate'])
mkt_info_frame.columns = ['tradeDate','code','ret','to']
mkt_info_frame.head()
```

```python
mkt_info_frame = mkt_info_frame.dropna()
```

```python
# 构造下一个月的收益率以及三个因子
all_signal = pd.DataFrame()

for i in range(1,len(month_end)):
    
    #先构造下一个月的月度收益率
    #获取几个关键性的日期
    begin = month_start[i-1]  #该月的第一天
    start = month_start[i]    #下一个月的第一天
    end = month_end[i]        # 该月的最后一天
    if i < len(month_end)-1:
        next_end = month_end[i+1]   #下一个月的最后一天
    else:
        next_end = '20180629'
    
    #获取研究改节点的股票代码，下一个月第一天的代码集合
    code = all_stock[all_stock['date']==start]['code'].tolist()
    
    #这个月月底的收盘价  后复权
    inf1 = DataAPI.MktEqudAdjAfGet(secID=code,tradeDate=end,isOpen="",\
                                   field=u"secID,tradeDate,closePrice",pandas="1")
    inf1.columns = ['code','tradeDate1','closePrice1']
    #下一个月月底的收盘价
    inf2 = DataAPI.MktEqudAdjAfGet(secID=code,tradeDate=next_end,isOpen="",\
                                   field=u"secID,tradeDate,closePrice",pandas="1")
    inf2.columns = ['code','tradeDate2','closePrice2']
    
    #两个进行合并
    inf= pd.merge(inf1,inf2,on ='code',how = 'outer')
    inf['month_ret'] = inf['closePrice2']/inf['closePrice1']-1
    inf['date'] = end
    inf = inf[['date','code','month_ret']]  #获得下月的股票月度收益率 日期标记为本月最后一天
    
    #构造三个因子
    period_frame = mkt_info_frame[(mkt_info_frame['tradeDate']>=begin)&(mkt_info_frame['tradeDate']<=end)]
    period_frame = period_frame[period_frame['code'].isin(code)]  #.isin()的用法
    period_frame['next_return']  = period_frame.groupby('code')['ret'].shift(-1)
    period_frame['abs_next_return'] = abs(period_frame['next_return'])
    period_frame = period_frame.dropna()
    
    #FR因子
    trading_date = [x for x in trade_list if x>=begin and x<=end]
    n_days = len(trading_date)/2
    signal_fr = period_frame.groupby('code').apply(lambda df:df['to'].corr(df['abs_next_return'],method = 'pearson',min_periods=n_days))
    
    #FRu因子
    period_frame_positive = period_frame[period_frame['next_return']>0]
    signal_fru = period_frame_positive.groupby('code').apply(lambda df:df['to'].corr(df['next_return'],method = 'pearson',min_periods=5))
    
    #FRd因子
    period_frame_negative = period_frame[period_frame['next_return']<=0]
    signal_frd = period_frame_negative.groupby('code').apply(lambda df:df['to'].corr(df['next_return'],method = 'pearson',min_periods=5))
 
    #联合
    signal = pd.concat([signal_fr,signal_fru],axis = 1)
    signal = pd.concat([signal,signal_frd],axis = 1)
    signal.columns = ['fr','fru','frd']
    signal['code']=signal.index
    signal['date'] = end
    signal = signal[['date','code','fr','fru','frd']]
    signal.sort_values(by='code',inplace = True)
    signal.reset_index(drop = True,inplace=True)
    
    #与inf进行联合
    signal=pd.merge(inf,signal,on=['date','code'],how='right')
    
    all_signal=all_signal.append(signal)
    
all_signal
```

```python
all_signal.to_csv('FR_factor/all_signal.csv', index=True)  # 保存数据，下一次继续使用节约时间
```

```python
#定义生成IC值的函数
import numpy as np
import scipy.stats as st
def cal_ic(data,factor_name):   #factor_name 本身就是str
    IC = []
    Month_end = sorted(list(set(data['date'].tolist())))   #注意此时的Month_end和前面生成的month_end不一样，int  str
    for date in Month_end:
        
        tdata = data[data['date']==date]       #  必须使用新名称，注意循环
        
        #tdata.sort_values(by='code', inplace=True)  #不排序也可以，不影响结果
        #tdata.reset_index(drop=True, inplace=True)
        
        factor_array = np.array(tdata[factor_name])
        return_array = np.array(tdata['month_ret'])        #相关系数生成的方法
        
        #cor,pval = st.spearmanr(factor_array,return_array,nan_policy = 'omit')
        cor = tdata[factor_name].corr(tdata['month_ret'],method = 'spearman')
        IC.append(cor)
        
    IC_data = pd.DataFrame(dict({'date':Month_end,'IC':IC}))  #生成字典，字典再转化为df
    IC_data = IC_data[['date','IC']]
    
    return IC_data  # 返回的为DF

#分别计算三个因子的IC值
raw_IC_fr = cal_ic(all_signal1,'fr')
raw_IC_fr.columns = ['date','IC_fr']

raw_IC_fru = cal_ic(all_signal1,'fru')
raw_IC_fru.columns = ['date','IC_fru']

raw_IC_frd = cal_ic(all_signal1,'frd')
raw_IC_frd.columns = ['date','IC_frd']

raw_IC =pd.merge(raw_IC_fr,raw_IC_fru,on='date')
raw_IC =pd.merge(raw_IC,raw_IC_frd,on='date')

raw_IC
```

```python
#计算三个因子的IC均值以及ICIR值
#FR
FR_IC = raw_IC['IC_fr'].mean()
FR_ICIR = raw_IC['IC_fr'].mean()/raw_IC['IC_fr'].std()*np.sqrt(12)
print('FR因子的IC和ICIR分别是：')
print(FR_IC,FR_ICIR)

#FRu
FRu_IC = raw_IC['IC_fru'].mean()
FRu_ICIR = raw_IC['IC_fru'].mean()/raw_IC['IC_fru'].std()*np.sqrt(12)
print('FRu因子的IC和ICIR分别是：')
print(FRu_IC,FRu_ICIR)

#FRd
FRd_IC = raw_IC['IC_frd'].mean()
FRd_ICIR = raw_IC['IC_frd'].mean()/raw_IC['IC_frd'].std()*np.sqrt(12)
print('FRd因子的IC和ICIR分别是：')
print(FRd_IC,FRd_ICIR)
```

```python
import statsmodels.api as sm
import pandas as pd
#第二部分，计算纯净因子的IC值
def get_industry(date,code):
    indu = DataAPI.MdSwBackGet(secID=code,field=u"secID,isNew,industryName1",pandas="1")
    indu = indu[indu['isNew']==1]
    indu.drop_duplicates(subset =['secID'],inplace = True)
    indu['date'] = date
    indu = indu[['date','secID','industryName1']]
    indu.columns = ['date','code','industryName']
    return indu
#code = ['600687.XSHG','000001.XSHE','600387.XSHG','600769.XSHG','600393.XSHG','000988.XSHE']    
#get_industry('20070401',code) 

#计算纯净因子

all_prue_fr = pd.DataFrame()
for date in month_end[1:]:
    
    #原计算出来的三个因子
    tdf =all_signal[all_signal['date']==date]
    code = list(tdf['code'])
    
    # 从平台上，获取其它的几个因子的数据
    factors = DataAPI.MktStockFactorsOneDayGet(tradeDate=date,secID=code,field=u"secID,tradeDate,LFLO,REVS20,VOL20,ATR14",pandas="1")
    factors['tradeDate'] = map(time_change,factors['tradeDate'])
    factors.columns = ['code','date','lflo','revs20','vol20','variance60']
    
    #获取行业的信息
    indu = get_industry(date,code)
    
    #三者链接起来
    tdf = pd.merge(tdf,factors,on=['date','code'])
    tdf = pd.merge(tdf,indu,on=['date','code'])
    tdf.dropna(inplace = True)
    
    
    # 对各因子进行标准化和去极值化
    tdf.set_index('code',inplace = True)
    for col in tdf.columns[2:-1]:
        temp = tdf[col]
        tdf[col] = standardize(winsorize(temp))
    
    #行业转化为哑变量
    indu_dummies = pd.get_dummies(tdf['industryName'])
    
    tdf.drop('industryName',axis = 1,inplace = True)
    
    #行业与tdf结合一起
    tdf = pd.merge(tdf,indu_dummies,left_index = True,right_index=True,how='outer')
    
    #对传统的三个因子做回归
    y1 = np.array(tdf['fr'])
    y2 = np.array(tdf['fru'])
    y3 = np.array(tdf['frd'])
    
    X = np.array(tdf[tdf.columns[5:]])
    X = sm.add_constant(X,has_constant='add')
    
    #回归，求prue_fr
    model = sm.OLS(y1,X,missing = 'drop')
    results = model.fit()
    tdf['fr'] = results.resid
    tdf['fr'] = standardize(winsorize(tdf['fr']))

    #回归，求prue_fru
    model = sm.OLS(y2,X,missing = 'drop')
    results = model.fit()
    tdf['fru'] = results.resid
    tdf['fru'] = standardize(winsorize(tdf['fru']))
    
    #回归，求prue_frd
    model = sm.OLS(y3,X,missing = 'drop')
    results = model.fit()
    tdf['frd'] = results.resid
    tdf['frd'] = standardize(winsorize(tdf['frd']))
    
    #对tdf获取需要的列，去掉没有用的列
    tdf['code'] = tdf.index
    tdf.reset_index(drop=True,inplace =True)
    tdf = tdf[['date','code','month_ret','fr','fru','frd']]
    tdf = tdf.sort_values(by=['date','code'])
    
    #增加到整个DF中
    all_prue_fr = all_prue_fr.append(tdf)
    

all_prue_fr
```
