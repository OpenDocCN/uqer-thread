```python
# 转换stockcode代码为str类型
def code2str(x):
    x_str = str(x)
    while int(x)<100000:
        x = x*10
        x_str = '0' + x_str
    return x_str
```

```python
# 提取交易日期对应的可交易的ykcode用以提取可转债价格信息，可能会重复，因为会同时发布两个可转债，要去重
def get_ykcode(df):
    date_list = list(set(df.tradeDate.values))
    ykcode_list = []
    for date in date_list:
        ykcode = list(set(df.ykcode[df.tradeDate == date]))
        ykcode_list.append(ykcode)
    new_df = pd.DataFrame({'date':date_list,'stockcode':ykcode_list})
    return new_df
```

```python
# 提取交易日期对应的可交易的stockcode，可能会重复，因为会同时发布两个可转债，要去重
def get_stockcode(df):
    date_list = list(set(df.tradeDate.values))
    stockcode_list = []
    for date in date_list:
        stockcode = list(set(df.stockcode[df.tradeDate == date]))
        stockcode_list.append(stockcode)
    new_df = pd.DataFrame({'date':date_list,'stockcode':stockcode_list})
    return new_df
```

```python
# get_startend生成开始-结束日期，为季度数据提取做准备
# 生成标准日期格式再匹配
# 有可转债的中文名字可能会重复，但是ykcode不会重复
# 去掉销售日期，get_cov不重复地获取所有债券名单
# 要生成values进行布尔索引，如果直接选取一列，会产生series，不能对由values生成的array进行布尔索引 ！！！
def get_cov(df):
    new_df = df.drop(['tradeDate'],axis = 1)
    new_df = new_df.drop_duplicates() 
    new_df = new_df.reset_index(drop = True)
    return new_df

def get_startend(df):   
    new_df = get_cov(df)
    df['tradeDate'] = df['tradeDate'].apply(lambda x: dt.datetime.strptime(x,'%Y/%m/%d'))
    new_df['start_date'] = new_df.apply(lambda x: min(df['tradeDate'].values[df.ykcode.values == x.ykcode]),axis = 1)
    # ykcode后面要加values，否则返回的是series，非array，没有办法根据布尔值进行索引
    new_df['end_date'] = new_df.apply(lambda x: max(df['tradeDate'].values[df.ykcode.values == x.ykcode]),axis = 1)
    return new_df
```

```python
start_end_df = get_startend(original_df)
```

# ifoperateRato_con	
近2期年报扣非净利润（扣除非经常性损益后的净利润）均小于0或者最新财报与上一年年报扣非净利润均小于0
DataAPI.FdmtEeGet
获取扣非净利润，Q1-第一季报，S1-半年报，Q3-第三季报，CQ3-三季报（累计1-9月），A-年报。
NIncomeCut	float	扣除非经常性损益后净利润
用np.isnan()来判断是否为空！！！

```python
FdmtEeGet_df = DataAPI.FdmtEeGet(ticker=ticker,secID=u"",reportType=u''
                                 ,endDate=end_date,beginDate=begin_date,publishDateEnd=u"",publishDateBegin=u"",field=u"",pandas="1")
```

```python
new_df = FdmtEeGet_df[quater_basic_factor+['NIncomeCut','reportType']]
```

```python
new_df = new_df[pd.isnull(new_df.NIncomeCut) == False]
```

```python
new_df.head()
```

```python
# 将新产生的指标放在记入新列，以publishDate为关键词
# A和CQ3都是要判断上一年的'年报'是否小于0
def ifoperateRato_con(x):
    if np.isnan(x.NIncomeCut):
        return np.NAN
    elif x.NIncomeCut<0:
        last_year = str(int(x.endDate[:4])-1)+'-12-31' # 转换成上一年的年报endDate 
        temp1 = new_df.ticker == x.ticker 
        temp2 = new_df.endDate == last_year # 有不同的publishDate对应同一个endDate
        temp3 = new_df.NIncomeCut[temp1 & temp2]
        if len(temp3)==0:
            # print('nan')
            return np.NAN
        elif True in list(temp3<0):
            # print('NIC:'+str(x.NIncomeCut))
            # print('1: '+str(temp3))
            return True # 连续两年都小于0
        else:
            # print('NIC:'+str(x.NIncomeCut))
            # print('0: '+str(temp3))
            return False # 上一年不小于0
    else:
        return False # 本年不小于0
```

```python
new_df['ifoperateRato_con'] = new_df.apply(lambda x: ifoperateRato_con(x),axis = 1)
```

# ifoperatecash_cond
最近一期财报及上2期年报经营性现金流净额和小于0（除银行、非银金融、房地产）

```python
# NCFOperateA	float	经营活动产生的现金流量净额
FdmtCFGet = DataAPI.FdmtCFGet(ticker=ticker,secID=u"",reportType=u"",
                        endDate=end_date,beginDate=begin_date,publishDateEnd=u"",publishDateBegin=u"",endDateRep="",
                              beginDateRep="",beginYear="",endYear="",fiscalPeriod="",
                              field=quater_basic_factor+['NCFOperateA','reportType'],pandas="1")
FdmtCFGet.head()
```

```python
# 有publishdate不同，但是endDate相同的年报。。。只要有一个true就true吧
# 可能为空的值也要考虑
def ifoperatecash(x):
    if x.NCFOperateA > 0:
        return False
    else:
        last_year1 = str(int(x.endDate[:4])-1) + '-12-31'
        last_year2 = str(int(x.endDate[:4])-2) + '-12-31'
        temp1 = (FdmtCFGet.endDate == last_year1) & (FdmtCFGet.reportType == 'A') & (FdmtCFGet.ticker == x.ticker)
        temp2 = (FdmtCFGet.endDate == last_year2) & (FdmtCFGet.reportType == 'A') & (FdmtCFGet.ticker == x.ticker)
        temp3 = FdmtCFGet.NCFOperateA[temp1]
        temp4 = FdmtCFGet.NCFOperateA[temp2]
        if (len(temp3) == 0) | (len(temp4) == 0):
            return np.NAN
        else:
            return (True in list(temp3<0)) & (True in list(temp4<0)) # 要转化为list才可以查找true是否在其中
```

```python
FdmtCFGet['ifoperatecash_cond'] = FdmtCFGet.apply(lambda x: ifoperatecash(x),axis = 1)
```

```python
FdmtCFGet.to_csv('ifoperatecash_cond.csv', index=False, encoding='GBK')
```

```python
def cubic_root(nr):
    if nr<0:
        return -np.power(-nr, float(1)/3)
    else:
        return np.power(nr, float(1)/3)
```

```python
FdmtISQPITGet1['year'] = FdmtISQPITGet1['endDate'].apply(lambda x:int(x[:4]))
FdmtISQPITGet1['period'] = FdmtISQPITGet1['endDate'].apply(lambda x:int(x[5:7])).map({3:1,6:2,9:3,12:4})

FdmtISQPITGet1['Ac_Q_OpEarning1'] = FdmtISQPITGet1.apply(lambda x:FdmtISQPITGet1[(FdmtISQPITGet1.ticker==x.ticker)&(FdmtISQPITGet1.year==x.year-1)&(FdmtISQPITGet1.period==x.period)]['operateProfit'].values.tolist(),axis=1)
FdmtISQPITGet1['Ac_Q_OpEarning1'] = FdmtISQPITGet1['Ac_Q_OpEarning1'].apply(lambda x:np.nan if len(x)==0 else x[-1])
FdmtISQPITGet1['Ac_Q_OpEarning1'] = ((FdmtISQPITGet1['operateProfit']-FdmtISQPITGet1['Ac_Q_OpEarning1'])/FdmtISQPITGet1['Ac_Q_OpEarning1'])

FdmtISQPITGet1['Ac_Q_OpEarning'] = FdmtISQPITGet1.apply(lambda x:FdmtISQPITGet1[(FdmtISQPITGet1.ticker==x.ticker)&(FdmtISQPITGet1.year==x.year-1)&(FdmtISQPITGet1.period==x.period)]['Ac_Q_OpEarning1'].values.tolist(),axis=1)
FdmtISQPITGet1['Ac_Q_OpEarning'] = FdmtISQPITGet1['Ac_Q_OpEarning'].apply(lambda x:np.nan if len(x)==0 else x[-1])
FdmtISQPITGet1['Ac_Q_OpEarning'] = ((FdmtISQPITGet1['Ac_Q_OpEarning1']-FdmtISQPITGet1['Ac_Q_OpEarning'])/FdmtISQPITGet1['Ac_Q_OpEarning1'])

FdmtISQPITGet1.tail()
```

```python
FdmtISQPITGet1.drop(['secID','partyID','operateProfit','year','period','Ac_Q_OpEarning1'],inplace=True,axis =1)
```

```python
FdmtISQPITGet1 = FdmtISQPITGet1[pd.isnull(FdmtISQPITGet1.Ac_Q_OpEarning)== False]
```
