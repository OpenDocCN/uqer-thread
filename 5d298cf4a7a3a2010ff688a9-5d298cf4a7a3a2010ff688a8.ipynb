```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import datetime as dt
from CAL.PyCAL import * # CAL provides classes and functions dealing with date, time, and calendars. 

sns.set()
pd.set_option('display.width', 200)
cal = Calendar('China.SSE')
```

```python
index = '000300'
data = DataAPI.MktIdxdGet(ticker = index, beginDate='20180101',endDate='20181201') # 指数日行情
data.index = data.tradeDate.apply(lambda x: dt.datetime.strptime(x, '%Y-%m-%d'))

data2 = DataAPI.MktIdxdGet(ticker =index, beginDate='20140101',endDate='20141201')
data2.index = data2.tradeDate.apply(lambda x: dt.datetime.strptime(x, '%Y-%m-%d'))

data['2006 - 2008'] = data['closeIndex']
data = data[['2006 - 2008']]
data['2014 - 2015'] = np.nan
data['2014 - 2015'][:len(data2.closeIndex.values)] = data2['closeIndex'].values
data = data[['2006 - 2008', '2014 - 2015']]

data.plot(figsize=(25,7))

data
```
