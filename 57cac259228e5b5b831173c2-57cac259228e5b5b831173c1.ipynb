# 如何用matplotlib绘制K线图
突然想画画K线图，谷歌了一下`python`的`matplotlib`包中的`finance`就可以画，研究了好一会终于可以大致画出自己想要的啦，和大家分享一下。

首先是头文件，主要是用`candlestick_ohlc`来画K线图

```python
import matplotlib.pyplot as plt
from matplotlib.finance import candlestick_ohlc
import matplotlib.dates as mdates
import datetime 
```

接着我们处理一下行情数据，查阅下`candlestick_ohlc()`函数可以看出，需要的输入是`time`、`open`、`hgih`、`low`、`close`组成的一个`tuple`。而且`time`的格式必须是数，所以得用时间戳(可以用`matplotlib.dates`中的`mdaters`进行转化)，下面以`600000`浦发银行为例，处理一下，行情数据。

```python
df_history = DataAPI.MktEqudAdjGet(secID=u"600000.XSHG",ticker=u"",beginDate=u"20160701",endDate=u"",isOpen="",
                   field=u"tradeDate,openPrice,highestPrice,lowestPrice,closePrice",pandas="1")
df_history = df_history[df_history['openPrice'] != 0]  ## 去除停牌的行情
df_history['tradeDate'] = df_history['tradeDate'].map(lambda x:mdates.date2num(datetime.datetime.strptime(x,'%Y-%m-%d')))
df_history['ii'] = range(len(df_history))
date = df_history['ii'].tolist()
ohlc = df_history['ii'].map(lambda x:tuple(df_history.iloc[x][['tradeDate','openPrice','highestPrice','lowestPrice','closePrice']])).tolist()
#weekday_ohlc = [tuple([i]+list(item[1:])) for i,item in enumerate(ohlc)]
```

```python
ohlc[0:5]
```

可以看到经过以上处理的行情数据变为一个`list of tuple`。接下来我们就可以进行画图了。

```python
fig = plt.figure(figsize=(18,7))
ax = plt.subplot(1,1,1)
candlestick_ohlc(ax, ohlc, width=0.6, colorup='r', colordown='g')
plt.show()
```

啊，这个结果还不错，但是还是有点丑啊，而且时间轴也好奇怪。我们要把`x`轴变成时间。

```python
fig = plt.figure(figsize=(18,7))
ax = plt.subplot(1,1,1)
candlestick_ohlc(ax, ohlc, width=0.6, colorup='r', colordown='g')
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
plt.show()
```

我们加入了``ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))``,将时间轴的格式转成了时间的模样，大功告成哈！

不过仔细一看，不对为啥有很多空隙呢，原来是那些空的都是非交易日，或者那只股票那些天停牌了，看起来很不舒服啊，通常的行情软件也不会留空隙，所以我们也要把空隙去掉。

试了好多办法都失败了啊，上`stack Overflow`搜了一圈，发现只有自己画一遍`x`轴了。首先转不转时间戳无所谓了，先就从0到时间长度，然后再设置一遍`xticklabels`就可以了。

```python
df_history = DataAPI.MktEqudAdjGet(secID=u"600000.XSHG",ticker=u"",beginDate=u"20160701",endDate=u"",isOpen="",
                   field=u"tradeDate,openPrice,highestPrice,lowestPrice,closePrice",pandas="1")
df_history = df_history[df_history['openPrice'] != 0]  ## 去除停牌的行情
df_history['tradeDate'] = df_history['tradeDate'].map(lambda x:mdates.date2num(datetime.datetime.strptime(x,'%Y-%m-%d')))
df_history['ii'] = range(len(df_history))
date = df_history['ii'].tolist()
ohlc = df_history['ii'].map(lambda x:tuple(df_history.iloc[x][['tradeDate','openPrice','highestPrice','lowestPrice','closePrice']])).tolist()
weekday_ohlc = [tuple([i]+list(item[1:])) for i,item in enumerate(ohlc)]
```

```python
weekday_ohlc[0:5]
```

```python
fig = plt.figure(figsize=(18,7))
ax = plt.subplot(1,1,1)
candlestick_ohlc(ax, weekday_ohlc, width=0.6, colorup='r', colordown='g')
ax.set_xticks(range(0,len(weekday_ohlc),5))### 每五天标一个日期
ax.set_xticklabels([mdates.num2date(ohlc[index][0]).strftime('%Y-%m-%d') for index in ax.get_xticks()])####
plt.xlim(0,len(weekday_ohlc)-1) #设置一下x轴的范围
plt.setp(plt.gca().get_xticklabels(), rotation=45, horizontalalignment='right') #将x轴的label转一下，好看一点
plt.show()
```

OK，大功告成，我们包成函数吧！

```python
import matplotlib.pyplot as plt
from matplotlib.finance import candlestick_ohlc
import matplotlib.dates as mdates
import datetime 
from dateutil.parser import parse

def get_candlestick_plot(ticker,start_date,end_date='-'):
    start_date = parse(start_date).strftime('%Y%m%d')
    end_date = parse(end_date).strftime('%Y%m%d')
    df_history = DataAPI.MktEqudAdjGet(secID=u"",ticker=ticker,beginDate=start_date,endDate=end_date,isOpen="",
                   field=u"tradeDate,openPrice,highestPrice,lowestPrice,closePrice",pandas="1")
    df_history = df_history[df_history['openPrice'] != 0]  ## 去除停牌的行情
    df_history['tradeDate'] = df_history['tradeDate'].map(lambda x:mdates.date2num(datetime.datetime.strptime(x,'%Y-%m-%d')))
    df_history['ii'] = range(len(df_history))
    date = df_history['ii'].tolist()
    ohlc = df_history['ii'].map(lambda x:tuple(df_history.iloc[x][['tradeDate','openPrice','highestPrice','lowestPrice','closePrice']])).tolist()
    weekday_ohlc = [tuple([i]+list(item[1:])) for i,item in enumerate(ohlc)]
    
    fig = plt.figure(figsize=(len(weekday_ohlc)*10.0/46,7))
    ax = plt.subplot(1,1,1)
    candlestick_ohlc(ax, weekday_ohlc, width=0.6, colorup='r', colordown='g')
    ax.set_xticks(range(0,len(weekday_ohlc),5))### 每五天标一个日期
    ax.set_xticklabels([mdates.num2date(ohlc[index][0]).strftime('%Y-%m-%d') for index in ax.get_xticks()])####
    ax.legend([ticker],frameon=False)
    plt.xlim(0,len(weekday_ohlc)-1) #设置一下x轴的范围
    plt.setp(plt.gca().get_xticklabels(), rotation=45, horizontalalignment='right') #将x轴的label转一下，好看一点
    plt.show()
```

```python
get_candlestick_plot('600519','2015-12-30')   #不写end_date默认今天
```
