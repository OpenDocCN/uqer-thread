# 通联杯2015量化策略大赛参赛方法

> 本文介绍了“通联杯”2015量化策略大赛参赛方法


### **1. 如何报名参加大赛？**

----

以下为大赛报名流程概要：

![报名](http://ww4.sinaimg.cn/large/78528393gw1es9e8k537sj20m808cwfj.jpg)

* #### **未注册**

  如果您还未注册“通联量化实验室”可以通过如下步骤报名参赛：
  
  1) 进入 [大赛首页](https://app.wmcloud.com/mercury/contest/) ，点击“我要报名”按钮。
  
  ![我要报名](http://ww2.sinaimg.cn/large/0066RikZgw1es9hi9ixd6j30el09ogn1.jpg)
  
  2) 你会进入注册页面，完成注册。
  
  3) 注册成功后，您会进入报名信息填写页面，您需要填写团队名称、参赛人员信息，参赛团队由1-3人组成, 2015年6月19日之后终止修改队员信息。
  
  ![报名信息](http://ww2.sinaimg.cn/large/0066RikZgw1es9hj1daxoj30hy0750t8.jpg)
  
  4) 点击确认报名后系统会提示您报名成功。


  
* #### **已注册**
  
  如果您已经注册“通联量化实验室”,但是还未报名参赛，可以登录后，直接进入 [大赛首页](https://app.wmcloud.com/mercury/contest/) ，点击“我要报名”按钮，填写报名信息后确认参赛。

### **2. 报名成功后如何参加比赛？**

----

报名成功后，您需要提交策略正式参赛，以下为报名成功后的参赛流程：

![报名成功](http://ww1.sinaimg.cn/large/78528393gw1es9ecxgqayj20m80dwwfu.jpg)

* #### **开始研究** 

  您可以直接去 **开始研究** 页面，在Notebook中研究策略，运行回测成功后，点击参加大赛按钮即可参赛

  1）进入 [开始研究](https://app.wmcloud.com/mercury/labs/)
  
  2）新建一个Notebook
  
  ![新建Notebook](http://ww2.sinaimg.cn/large/0066RikZgw1es9hrh73z5j30c204xjrv.jpg)
  
  3）将单元模式切换为Strategy
  
  ![strategy模式](http://ww3.sinaimg.cn/large/0066RikZgw1es9hxfoztqj30cc062jrz.jpg)
  
  4）运行单元，显示策略回测结果
  
  5）单击参加大赛按钮，将该策略提交参赛
  
  ![参加大赛](http://ww1.sinaimg.cn/large/0066RikZgw1es9i00w0yoj30kx05laau.jpg)
  
  如果您还未报名参赛，将不会出现参加大赛按钮，请参考“1. 如何报名参加大赛？”内容报名参赛。
  
  6）进入模拟交易页面，如果您提交策略的大赛状态为“参赛中”，即表明参赛成功
  
  ![参赛中](http://ww4.sinaimg.cn/large/0066RikZgw1es9id64kcpj30dw05lwex.jpg)

* #### **社区** 
  
  您也可以通过**社区**学习、挖掘策略灵感。
  
  1）克隆策略至开始研究页面，例如您可以打开 [一个社区分享的策略](https://app.wmcloud.com/mercury/community/share/5559ce71f9f06c6c7504f8a9) 进行克隆：
  
  ![克隆](http://ww2.sinaimg.cn/large/0066RikZgw1es9i8mq6gqj30m503xaac.jpg)
  
  2）在策略中添加自己的灵感，编写相应的算法
  
  3）运行回测成功后，点击参加大赛按钮即可参赛。
  

### **3. 如何编写一个量化策略？**

----

* 如何开始学习量化知识？
  
  **社区** 中 **新手专栏** 提供了深入浅出的，从学习Python到学习量化分析的系列学习文档： [社区](https://app.wmcloud.com/mercury/community/)
  
* 如何在 **开始研究** 中写出一个策略？
  
  **开始研究** 是一个强大的量化策略研究与实现平台，参考帮助文档中介绍了它的详细使用方法：  [如何写个策略](https://app.wmcloud.com/mercury/help/faq/#如何写个策略)


### **4. 如何查看我的参赛策略**

  * 您可以在 [模拟交易](https://app.wmcloud.com/mercury/trade/) 页面查看当前参赛策略的模拟交易信息；
  
  * 您可以在 [大赛排名](https://app.wmcloud.com/mercury/contest/rank) 页面查看比赛前五十名的排名信息，以及您当前参赛策略的排名信息。
