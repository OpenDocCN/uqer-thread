除了某一天的截面数据，DataAPI也支持获取过去一段时间的数据，通过beginDate与endDate参数即可方便的获取。如图，我们可以调取平安银行与浦发银行从2020年6月1日到6月9日的收盘数据和总市值数据

```python
DataAPI.MktEqudGet(tradeDate=u"",secID=u"",ticker=["000001","600000"],
                   beginDate='20200601',endDate='20200609',
                   isOpen="",field=u"secID,secShortName,tradeDate,closePrice,marketValue",pandas="1")
```

股票日行情数据DataAPI支持提取多只股票再某一时间段的数据，我们再来看看量化投资常用的因子数据。

```python
如果想调用一只股票在某一时间段的因子数据，则应该使用另一个DataAPI
```

```python
DataAPI.MktEqudGet(secID=["000001.XSHE","600000.XSHG"],ticker=u"",tradeDate=u"20200609",
                   isOpen="",field=u"secID,secShortName,tradeDate,closePrice,marketValue",pandas="1")
```

```python
#调用某一天，多只股票的因子数据
#DataAPI.MktStockFactorsOneDayProGet(tradeDate=u"20170612",secID=u"",ticker=u"000001,600000",field=u"",pandas="1")
DataAPI.MktStockFactorsOneDayGet(tradeDate=u"20200609",secID=["600000.XSHG","000001.XSHE"],ticker=u"",field=u"secID,tradeDate,PS,PB,EPS,NetProfitGrowRate",pandas="1")
```

```python
#调用1只股票，某一时间段的因子数据
#DataAPI.MktStockFactorsDateRangeGet(secID=u"",ticker=u"000001",beginDate=u"20170612",endDate=u"20170616",field=u"",pandas="1")
DataAPI.MktStockFactorsDateRangeGet(secID=u"600000.XSHG",ticker=u"",beginDate=u"20200601",endDate=u"20200609",
                                    field=u"secID,tradeDate,PS,PB,EPS,NetProfitGrowRate",pandas="1")
```

如果我们还是想取多只股票再某一时间段的因子数据，则可以写循环来多次调用DateAPI,将它们全部取用出来

```python
import pandas as pd
df=pd.DataFrame()
for ticker in ["600000","000001"]:
    tmp_df=DataAPI.MktStockFactorsDateRangeGet(secID=u"",ticker=ticker,beginDate=u"20200601",
                                               endDate=u"20200609",
                                    field=u"secID,tradeDate,PS,PB,EPS,NetProfitGrowRate",pandas="1")
    
    #concat数据的合并,默认是纵向扩展，axis=1表示横线扩展
    df=pd.concat([df,tmp_df],axis=1) 
df.reset_index(drop=True)
```
