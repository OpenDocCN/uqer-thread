## 1.1 确定universe

```python
# coding: utf-8

import numpy as np
import pandas as pd
import datetime
import time
import os
import copy
import matplotlib.pyplot as plt
import statsmodels.api as sm
import scipy.stats as st
from CAL.PyCAL import *    # CAL.PyCAL中包含font
universe = set_universe('A')
cal = Calendar('China.SSE')

def time_change(x):
    y = datetime.datetime.strptime(x, '%Y-%m-%d')
    y = y.strftime('%Y%m%d')
    return y

# 获取回测区间的交易日、月末以及月初时间
def get_trade_list(start_date, end_date):
    """
    Args:
        start_date: 时间区间起点
        end_date: 时间区间终点
    Returns: 
        trade_list: 时间区间内的交易日
        month_end: 月末时间
        month_start: 月初时间
    """
    cal_dates = DataAPI.TradeCalGet(exchangeCD=u"XSHG", beginDate=start_date, endDate=end_date).sort('calendarDate')
    cal_dates = cal_dates[cal_dates['isOpen'] == 1]
    trade_list = cal_dates['calendarDate'].values.tolist()
    trade_list = [time_change(x) for x in trade_list]
    month_end = cal_dates[cal_dates['isMonthEnd'] == 1]
    month_end = month_end['calendarDate'].values.tolist()
    month_end = [time_change(x) for x in month_end]
    cal = Calendar('China.SSE')
    month_start = [cal.advanceDate(x, '1B').strftime('%Y%m%d') for x in month_end]
    return trade_list, month_end, month_start


# 获取股票池函数
def get_stock_pool(date):
    """
    Args:
        date: 月初时间
        N: 新股的定义时间
    Returns:
        stock_pool: 此月初的股票池
    """
    univ=DynamicUniverse('ZZ500')+DynamicUniverse('HS300')
    secids = univ.preview(date,skip_halted=False)
    df = pd.DataFrame({'secID': secids})
    df['trade_date'] = date
    df = df[['trade_date', 'secID']]
    return df

# 获取每月初的股票池(20060428 - 20180330)
tic = time.time()
start_date = '20070601'
end_date = '20180531'        
trade_list, month_end, month_start = get_trade_list(start_date, end_date)
all_stock = pd.DataFrame()
secid_list = []
for date in month_start:
    stock_pool = get_stock_pool(date)
    secid_list.extend(list(stock_pool.secID.values))

secID_pool = list(set(secid_list))
len(secID_pool)
```

## 1.2 取因子数据，并对数据进行标准化

```python
# coding: utf-8

import pandas as pd
import numpy as np
import os
import time
import multiprocessing
from multiprocessing.dummy import Pool as ThreadPool

# # 因子文件存放目录，如果目录不存在，程序自动新建一个
raw_data_dir = "./data_folder"
if not os.path.exists(raw_data_dir):
    os.mkdir(raw_data_dir)

# 定义因子
fields = ['ROE', 'QuickRatio', 'ETOP', 'TVMA20', 'REVS20','negMarketValue']


# 取原始因子数据
start ='2007-06-01'
end = '2018-05-31'



def get_factor_by_day(tdate):
    '''
    根据日期，获取当天的因子值
    tdate格式为：%Y%m%d
    '''
    cnt = 0
    while True:
        try:
            x = DataAPI.MktStockFactorsOneDayProGet(tradeDate=tdate,secID=secID_pool,field=['secID', 'tradeDate'] + fields,pandas="1")
            return x
        except Exception as e:
            cnt += 1
            if cnt >= 3:
                print('error get factor data: ', tdate)
                print e
                break

print("begin to get factor value for each stock...")

# 拿到交易日历，得到月末日期
trade_date = DataAPI.TradeCalGet(exchangeCD=u"XSHG", beginDate=start.replace("-", ""), endDate=end.replace("-",""), field=u"", pandas="1")
trade_date = trade_date[trade_date.isOpen == 1]
# # 取得每个月末日期，所有股票的因子值
pool = ThreadPool(processes=16)
date_list = [tdate.replace("-", "") for tdate in trade_date.calendarDate.values]
frame_list = pool.map(get_factor_by_day, date_list)
pool.close()
pool.join()
print "factor value get finished!"

factor_data = pd.concat(frame_list, axis=0)
factor_data.rename(columns={"tradeDate":"trade_date"}, inplace=True)

# 取个股的流通市值
neg_frame = DataAPI.MktEqudGet(secID=secID_pool,beginDate=start,endDate=end,field=[u"tradeDate", u'secID', u'negMarketValue'],pandas="1")
neg_frame.rename(columns={"tradeDate":"trade_date"}, inplace=True)

factor_data = factor_data.merge(neg_frame, on=['trade_date', 'secID'], how='inner')
# 对因子进行标准化
# 根据截面上的因子值大小进行排序标准化，因子值越大，标准化后的值也越大
def factor_standarize(df):
    tmp_df = df.copy()
    df[fields] = df[fields].rank()/len(df)
    return df

factor_data.dropna(subset=fields, how='all')
factor_frame = factor_data.groupby(['trade_date']).apply(factor_standarize)
factor_frame.to_pickle(os.path.join(raw_data_dir, "factors_6.pickle"))
factor_frame.head()
```

## 1.3 取外部变量因子数据，作为模型特征的一部分

```python
# 市场变量
idx = ['000300.ZICN', '000905.ZICN']
idx_fields=['closePrice','turnoverVol']
start ='2007-10-01'
end = '2018-05-31'

idx_data = get_data_cube(idx, idx_fields, start, end, style='ast')
idx_data['chpct'] = idx_data['closePrice'].copy()
idx_data['chpct'] = idx_data['chpct'].shift(20)
idx_data['chpct'] = (idx_data.closePrice - idx_data.chpct) / idx_data.chpct
# 涨跌幅做一阶差分
idx_data['shift_chpct_1'] = idx_data['chpct'].shift(1)
idx_data['chpct_diff'] = idx_data['chpct'] - idx_data['shift_chpct_1']

#月波动率因子
idx_data['volatility20'] = idx_data['closePrice'].rolling(20, min_periods=1).std()

# 月均换手量
idx_data['tv20'] = idx_data['turnoverVol'].rolling(20, min_periods=1).mean()
# 计算变化率
idx_data['shift_tv20_1'] = idx_data['tv20'].shift(1)
idx_data['tv20_chpct'] = (idx_data['tv20'] - idx_data['shift_tv20_1'])/idx_data['shift_tv20_1']

# 转换数个格式
idx_data = idx_data[['chpct_diff', 'volatility20', 'tv20_chpct']]
idx_data = idx_data.to_frame().reset_index()
idx_data.rename(columns={"major":"trade_date"}, inplace=True)
idx_data.set_index('trade_date', inplace=True)
idx_data_300=idx_data[idx_data.minor=='000300.ZICN']
del idx_data_300['minor']
idx_data_500 = idx_data[idx_data.minor=='000905.ZICN']
del idx_data_500['minor']

idx_data_300.columns = [x+"_HS300" for x in idx_data_300.columns]
idx_data_500.columns = [x+"_ZZ500" for x in idx_data_500.columns]

idx_data = pd.concat([idx_data_300, idx_data_500], axis=1)
idx_data.reset_index(inplace=True)
idx_data.to_pickle("outer_factor_market.pickle")



# 宏观变量

'''>>>>>>>>>>>>>>>>>>>>>>> 宏观数据, 月度数据部分 <<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<'''
# 取交易日历，将数据的trade_date变成每个月最后一个交易日
trade_calendar = DataAPI.TradeCalGet(exchangeCD=u"XSHG",beginDate=u"20070101",endDate=u"20170531",field=u"",pandas="1")
trade_calendar = trade_calendar[trade_calendar.isMonthEnd==1]
trade_calendar['month'] = trade_calendar['calendarDate'].apply(lambda x: "".join(x.split("-")[:2]))
trade_calendar['next-month'] = trade_calendar['month'].shift(-1)
trade_calendar = trade_calendar[['calendarDate', 'month', 'next-month']]


macro_frame = pd.read_csv(os.path.join(raw_data_dir, u'宏观数据.csv'), encoding='gbk', index_col=0)
macro_frame.reset_index(inplace=True)
macro_frame.rename(columns={"index":"trade_date"}, inplace=True)
macro_frame['month'] = macro_frame['trade_date'].apply(lambda x: "".join(x.split("-")[:2]))

# 对齐交易日历
macro_frame = macro_frame.merge(trade_calendar, on=['month'], how='left')
del macro_frame['trade_date']
del macro_frame['month']
macro_frame.rename(columns={"calendarDate":"trade_date"}, inplace=True)
macro_frame.sort_values(by=['trade_date'], inplace=True)

# 将日期移后一个月（数据披露会延迟一个月）
macro_frame['trade_date'] = macro_frame['trade_date'].shift(-1)
macro_frame['next-month'] = macro_frame['next-month'].shift(-1)
macro_frame.dropna(inplace=True)

macro_frame.set_index(['trade_date', 'next-month'], inplace=True)
# 计算一阶差分
macro_list1 = [u'CPI同比', u'M1同比', u'M2同比', u'PPI同比', u'PMI']
# 前值frame
pre_macro_frame = macro_frame[macro_list1].shift(1)
macro1_frame = macro_frame-pre_macro_frame
macro1_frame.reset_index(inplace=True)
# 数据的使用时间在下个月日期
macro1_frame.rename(columns={"next-month":"month"}, inplace=True)
del macro1_frame['trade_date']


'''>>>>>>>>>>>>>>>>>>>>>>> 宏观数据, 日度数据部分 <<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<'''
# shibor1w
shibor1w_frame =DataAPI.MktIborGet(ticker=u"Shibor1w",beginDate=u"20070601",endDate=u"20170531",currency=u"",field=[u'tradeDate', u'rate'],pandas="1")

# 美元对人民币汇率
currency_frame = DataAPI.MktFxRefRateGet(currencyPair=u"USD/CNY",beginDate=u"20070601",endDate=u"20170531",field=[u"tradeDate", 'midRate'],pandas="1")
macro2_frame = shibor1w_frame.merge(currency_frame, on=['tradeDate'], how='inner')
macro2_frame.rename(columns={"rate":"shibor1w", 'midRate':'usd-cny', 'tradeDate':"trade_date"}, inplace=True)
macro2_frame.head()
# 计算变化率
macro2_frame.set_index('trade_date', inplace=True)
macro2_frame = macro2_frame.pct_change()
macro2_frame.reset_index(inplace=True)
macro2_frame['month'] = macro2_frame['trade_date'].apply(lambda x: "".join(x.split("-")[:2]))


'''>>>>>>>>>>>>>>>>>>>>>>> 宏观数据, 合并两部分数据，最终数据为日度 <<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<'''
macro_frame = macro2_frame.merge(macro1_frame, on=['month'], how='left')
macro_frame.dropna(inplace=True)
del macro_frame['month']
macro_frame.to_pickle(os.path.join(raw_data_dir, 'outer_factor_macro.pickle'))

########################### 汇总外部变量数据 ##################################
# 市场变量 + 宏观变量 = 外部变量
outer_factor_frame = idx_data.merge(macro_frame, on=['trade_date'], how='inner')
outer_factor_frame.to_pickle(os.path.join(raw_data_dir, 'outer_factor.pickle'))
outer_factor_frame.head()
```

## 1.4 计算因子的IC， 用历史IC作为模型特征的一部分

```python
start_t = time.time()
# 取交易日历
trade_calendar = DataAPI.TradeCalGet(exchangeCD=u"XSHG",beginDate=u"20070101",endDate=u"20171231",field=u"",pandas="1")
trade_calendar = trade_calendar[trade_calendar.isOpen==1]
for n in [5, 20]:
    trade_calendar['pre%s_date'%n] = trade_calendar['calendarDate'].shift(n)
trade_calendar['next5_date'] = trade_calendar['calendarDate'].shift(-5)

trade_calendar.dropna(inplace=True)
trade_calendar = trade_calendar[['calendarDate', 'prevTradeDate', 'pre5_date', 'pre20_date', 'next5_date']]
trade_calendar.rename(columns={"prevTradeDate":'pre1_date'}, inplace=True)
trade_calendar.rename(columns={"calendarDate":"trade_date"}, inplace=True)


# 取行情数据
close_frame = get_data_cube(secID_pool, ['closePrice'], start, end, style='ast', adj='pre')['closePrice']
close_frame.reset_index(inplace=True)
close_frame = pd.melt(close_frame, id_vars='index', value_vars=[x for x in close_frame.columns if x !='index'])
close_frame.columns=['trade_date', 'secID', 'closePrice']

# 把收盘价根据N=[1,5,20]天前合并到因子的dataframe中，便于计算IC

factor_frame = factor_frame.merge(trade_calendar, on=['trade_date'], how='left')
factor_frame = factor_frame.merge(close_frame, on=['trade_date', 'secID'], how='left')
# 准备数据，用来计算上一期的IC，作为模型的feature
for n in [1, 5, 20]:
    # 合并前n日的收盘价
    merged_price_frame = close_frame.copy()
    merged_price_frame.rename(columns={"trade_date":"pre%s_date"%n, "closePrice":"pre%s_closePrice"%n}, inplace=True)
    factor_frame = factor_frame.merge(merged_price_frame, on=['secID', 'pre%s_date'%n], how='left')
    factor_frame['pre%s_chpct'%n] = (factor_frame['closePrice'] - factor_frame["pre%s_closePrice"%n])/factor_frame["pre%s_closePrice"%n]
    # 合并前n日的因子值
    merged_factor_frame = factor_frame[['trade_date', 'secID']+fields]
    rename_list = {}
    for factor_name in fields:
        rename_list[factor_name] = "pre%s_%s"%(n, factor_name)
    rename_list['trade_date']="pre%s_date"%n
    merged_factor_frame.rename(columns=rename_list, inplace=True)
    factor_frame = factor_frame.merge(merged_factor_frame, on=['secID', 'pre%s_date'%n], how='left')

# 准备数据，用来计算未来的IC，作为模型的预测目标(未来5天)
merged_price_frame = close_frame.copy()
merged_price_frame.rename(columns={"trade_date":"next5_date", "closePrice":"next5_closePrice"}, inplace=True)
factor_frame = factor_frame.merge(merged_price_frame, on=['secID', 'next5_date'], how='left')
factor_frame['next5_chpct'] = (factor_frame['next5_closePrice'] - factor_frame['closePrice'])/factor_frame['closePrice']
# 将基础数据存储下来，便于debug
factor_frame = pd.read_pickle(os.path.join(raw_data_dir, "factor_closeprice_details.pickle"))

''' >>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>> 开始计算前N期的IC，以及当前时间点因子的IC（和未来收益） <<<<<<<<<<<<<<<<<<<<'''
# ic计算方法为皮尔逊系数
def calc_ic(date_df):
    df = date_df.copy()
    result_dict={}
    for factor_name in fields:
        for n in [1,5, 20]:
            df["%s_pre_IC%s"%(factor_name, n)] = df['pre%s_%s'%(n, factor_name)].corr(df['pre%s_chpct'%n])
        df['%s_IC'%factor_name] = df[factor_name].corr(df['next5_chpct'])
    return df


# 计算上一期的IC和未来的IC
ic_frame = factor_frame.groupby(['trade_date']).apply(calc_ic).drop_duplicates(subset=['trade_date'])
ic_columns = ['trade_date'] + [x for x in ic_frame.columns if "_pre_IC" not in x and "_IC" in x ]+[x for x in ic_frame.columns if "_pre_IC" in x]
ic_model_frame = ic_frame[ic_columns]
ic_model_frame.index = range(len(ic_model_frame))

end_t = time.time()
print "time cost:%s"%(end_t -start_t)
# 将IC真实值、历史值存储下来
ic_model_frame.to_pickle(os.path.join(raw_data_dir, "ic_model_data.pickle"))
ic_model_frame.head()
```

## 2.2 用xgboost对因子IC进行预测

- static: 代表不进行滚动建模，2008-2013数据进行训练，用训练得到的模型预测因子值（6个模型）

- rolling: 代表滚动建模，从2014年开始，每一个季度后将训练数据滚动一个季度训练得到新的模型，并用在下一个季度的预测上

- 得到的static_result/rolling_result 为 dataframe，列为 trade_date, 【各个因子的实际IC】， 【各个因子的预测IC】

- xgboost有很多超参数可以调整，在本文中，我们对max_depth in [3, 4], eta in [0.01, 0.02, 0.1, 0.2]进行了调参，最终选择了 max_depth=3, eta=0.01

- 时间说明： 该部分耗时较长，跑static的模型，在30分钟左右，跑rolling模型的时间在1-2个小时

```python
import xgboost as xgb
import time
import pandas as pd
import numpy as np

class BoostModel:
    '''
    定义一个利用xgboost来做boosting tree的流程
    '''
    def __init__(self, max_depth=3, subsample=0.95, num_round=2000, early_stopping_rounds=50):
        '''
        初始化
        参数:
            max_depth：树的最大深度
            subsample： 采样率，也就是选取多少比例的样本进行训练
            num_round： 最大循环次数，也就是最大的树个数
            early_stopping_rounds：提前停止次数，假设为50，也就是说验证集的误差迭代到一定程度在50次内不能再继续降低，就停止迭代
        '''
        
        #下面的参数eta: 学习率；silent: 是否打印训练过程信息；alpha: L1 正则的惩罚系数；lambda: L2 正则的惩罚系数；
        # eval_metric: 评价指标；objective: 定义学习任务及相应的学习目标
        self.params = {'max_depth': 3, 'eta': 0.01, 'silent': 1, 'alpha': 0.5, 'lambda': 0.5, 'eval_metric':'rmse', 'subsample':subsample, 'objective': 'reg:linear', 'colsample_bytree':0.8}
        self.num_round = num_round
        self.early_stopping_rounds = early_stopping_rounds


    def fit(self, train_data, train_label,  val_data, val_label):
        '''
        训练模型
        参数:
            train_data, train_label：分别对应着训练特征数据，训练标签
            val_data, val_label： 分类对应着验证特征数据，验证标签
        返回:
            boost_model, 训练后的xgboost模型
        '''
        
        dtrain = xgb.DMatrix(train_data, label=train_label)
        deval = xgb.DMatrix(val_data, label=val_label)

        boost_model = xgb.train(self.params, dtrain, num_boost_round=self.num_round, evals=[(dtrain,'train'), (deval, 'eval')], early_stopping_rounds=self.early_stopping_rounds, verbose_eval=False)
        print('get best eval : %s, in step %s'%(boost_model.best_score, boost_model.best_iteration))
        self.boost_model = boost_model
        
        return boost_model

    def predict(self, test_data):
        '''
        预测
        参数:
            test_data：待预测的特征数据
        返回:     
            predict_score: 预测值
        '''
        
        dtest = xgb.DMatrix(test_data)
        predict_score = self.boost_model.predict(dtest, ntree_limit=self.boost_model.best_ntree_limit)
        
        return predict_score
```

```python
import xgboost as xgb
import time
import pandas as pd
import numpy as np
import os


class BoostModel:
    '''
    定义一个利用xgboost来做boosting tree的流程
    '''

    def __init__(self, max_depth=3, subsample=0.95, num_round=2000, early_stopping_rounds=50):
        '''
        初始化
        参数:
            max_depth：树的最大深度
            subsample： 采样率，也就是选取多少比例的样本进行训练
            num_round： 最大循环次数，也就是最大的树个数
            early_stopping_rounds：提前停止次数，假设为50，也就是说验证集的误差迭代到一定程度在50次内不能再继续降低，就停止迭代
        '''

        # 下面的参数eta: 学习率；silent: 是否打印训练过程信息；alpha: L1 正则的惩罚系数；lambda: L2 正则的惩罚系数；
        # eval_metric: 评价指标；objective: 定义学习任务及相应的学习目标
        self.params = {'max_depth': max_depth, 'eta': 0.02, 'silent': 1, 'alpha': 0.5, 'lambda': 0.5,
                       'eval_metric': 'rmse', 'subsample': subsample, 'objective': 'reg:linear',
                       'colsample_bytree': 0.8}
        self.num_round = num_round
        self.early_stopping_rounds = early_stopping_rounds

    def fit(self, train_data, train_label, val_data, val_label):
        '''
        训练模型
        参数:
            train_data, train_label：分别对应着训练特征数据，训练标签
            val_data, val_label： 分类对应着验证特征数据，验证标签
        返回:
            boost_model, 训练后的xgboost模型
        '''

        dtrain = xgb.DMatrix(train_data, label=train_label)
        deval = xgb.DMatrix(val_data, label=val_label)

        boost_model = xgb.train(self.params, dtrain, num_boost_round=self.num_round,
                                evals=[(dtrain, 'train'), (deval, 'eval')],
                                early_stopping_rounds=self.early_stopping_rounds, verbose_eval=False)
        # print('get best eval auc : %s, in step %s' % (boost_model.best_score, boost_model.best_iteration))
        self.boost_model = boost_model

        return boost_model

    def predict(self, test_data):
        '''
        预测
        参数:
            test_data：待预测的特征数据
        返回:
            predict_score: 预测值
        '''

        dtest = xgb.DMatrix(test_data)
        predict_score = self.boost_model.predict(dtest, ntree_limit=self.boost_model.best_ntree_limit)

        return predict_score

    

# 拆分训练、验证、测试集
def get_train_val_test_data(df, train_pct=0.8, train_start_date=None, train_end_date=None, test_start_date=None,
                            test_end_date=None):
    '''
    参数:
        df: 数据集，columns包括：trade_date(YYYY-mm-dd), 各个因子的真实IC，特征
        train_start_date, train_end_date: 时间区段内的数作为训练集
        train_pct: 训练集中，多大的比例作为验证集
    返回:
        train_df, val_df, test_df: 三个均为DataFrame，对应着训练、验证、测试的数据集
        上述3个dataframe的columns依然为 trade_date, 各个因子的真实IC，特征
    '''
    t1 = time.time()

    train_val_df = df[(df['trade_date'] >= train_start_date) & (df['trade_date'] <= train_end_date)]
    train_val_df = train_val_df.sample(frac=1, random_state=666).reset_index(drop=True)

    # 拆分训练集、验证集
    train_df = train_val_df.iloc[0:int(len(train_val_df) * train_pct)]
    val_df = train_val_df.iloc[int(len(train_val_df) * train_pct):]

    test_df = df[df.trade_date > train_end_date]
    if test_start_date and test_end_date:
        test_df = test_df[(test_df.trade_date >= test_start_date) & (test_df.trade_date <= test_end_date)]
    t2 = time.time()
    print("get train_val_test_data, time_cost:", t2 - t1)
    print("[data verify], train, %s to %s" % (min(train_df.trade_date.values), max(train_df.trade_date.values)))
    print("[data verify], val, %s to %s" % (min(val_df.trade_date.values), max(val_df.trade_date.values)))
    print("[data verify], test, %s to %s" % (min(test_df.trade_date.values), max(test_df.trade_date.values)))
    return train_df, val_df, test_df

# 将因子格式的dataframe，转换成feature和label两个array，便于输入模型
def format_feature_label(data_df, y_marker, factor_names):
    '''
    参数:
        df: 原始输入数据，DataFrame, columns包括：trade_date(可能没有), 各个因子的真实IC，特征
        y_marker: 真实值列的结尾标识  [各个因子的真实IC列: ROE_IC, GROW_IC, EP_IC等]
        factor_names: list, 作为label的因子列表，label_lists的顺序和factor_names严格一致
        除了trade_date, 带y_marker的列都会被认为是特征
    返回:
        feature: np.array, 对应着更改格式后的数据特征, N*K个特征
        label_lists: 单个元素为 np.array, 对应着更改格式后的数据标签, N*1
    '''
    df = data_df.copy()
    if 'trade_date' in df.columns:
        del df['trade_date']
    end_tag_length = len(y_marker)
    # 提取出特征列
    valid_feature_col = [x for x in df.columns if x[-end_tag_length:] != y_marker]
    # 特征数组
    feature = np.array(df[valid_feature_col])
    # 标签数组（预测的真实值组）
    label_lists = []
    for factor_name in factor_names:
        label = np.array(df[factor_name + y_marker])
        label_lists.append(label)
    return feature, label_lists

# 计算n个季度之后的日期
def calc_time(ori_date, periods):
    '''
    输入参数：
        ori_date: 开始日期，YYYY-mm-dd格式
        periods: 季度数
    输出：
        periods个季度之后的日期stop_date，以及stop的下一个日期next_start_date
    '''
    end_fix = {"01": "01", "04": "01", "07": "01", "10": "01", "03": "31", "06": "30", "09": "30", "12": "31"}
    [ori_y, ori_m, ori_d] = ori_date.split("-")

    ori_y = int(ori_y)
    ori_m = int(ori_m)
    next_start_m = ori_m + int(periods)
    start_year_skip = next_start_m / 12
    next_start_m = next_start_m % 12
    next_start_y = ori_y + start_year_skip
    stop_m = next_start_m
    stop_y = next_start_y

    if stop_m == 1:
        stop_m = 12
        stop_y -= 1
    else:
        stop_m -= 1
    next_start_date = "%s-%s-%s" % (next_start_y, str(next_start_m).zfill(2), end_fix[str(next_start_m).zfill(2)])
    stop_date = "%s-%s-%s" % (stop_y, str(stop_m).zfill(2), end_fix[str(stop_m).zfill(2)])

    return stop_date, next_start_date

# 划分时间区段
def get_rolling_date(train_start, test_start, test_end, test_quarter_periods=3, train_quarter_periods=72):
    '''
    1. 时间都是YYYY-mm-dd格式
    2. 从test_start到test_end，按照季度平移训练期， 分成一个个子区间
    test_quarter_periods = 3  # 测试期为3个月
    train_quarter_periods = 72 # 保持训练期为6年
    返回：
        [[train_s, train_e, test_s, test_e], [train_s, train_e, test_s, test_e]]
    '''
    test_begin = test_start
    train_begin = train_start

    date_constrains = []
    while 1:
        test_stop, next_test_start = calc_time(test_begin, test_quarter_periods)
        train_stop, _ = calc_time(train_begin, train_quarter_periods)
        _, next_train_start = calc_time(train_begin, test_quarter_periods)
        if test_stop >= test_end:
            test_stop = test_end
            date_constrains.append([train_begin, train_stop, test_begin, test_stop])
            break
        else:
            date_constrains.append([train_begin, train_stop, test_begin, test_stop])
        test_begin = next_test_start
        train_begin = next_train_start
    return date_constrains

# 用xgboost模型来预测factor_names中各个因子的IC值（对应的特征和时间在test_df中）
def predict_factors(train_df, val_df, test_df, factor_names):
    '''
    输入参数：
        train_df, val_df, test_df: 训练集、验证集、测试集数据
    输出：
        预测值df，columns为 trade_date, 因子真实值（这两项是取自test_df中），因子的预测值

    '''
    train_feature, train_labels = format_feature_label(train_df, "_IC", factor_names)
    val_feature, val_labels = format_feature_label(val_df, "_IC", factor_names)
    test_feature, test_labels = format_feature_label(test_df, "_IC", factor_names)

    test_date_values = list(test_df.trade_date.values)
    result_dict = {"trade_date": test_date_values}
    # xgboost模型训练，得到因子值输出
    for tcount in range(len(factor_names)):
        factor_name = factor_names[tcount]
        print("predicting %s..." % factor_name)
        if debug_flag:
            boost_model = BoostModel(num_round=200, early_stopping_rounds=20)
        else:
            boost_model = BoostModel()
        train_start = time.time()
        boost_model.fit(train_feature, train_labels[tcount], val_feature, val_labels[tcount])
        train_stop = time.time()
        print("train time cost:%s" % (train_stop - train_start))
        predict_score = boost_model.predict(test_feature)
        predict_stop = time.time()
        print("predict time cost:%s" % (predict_stop - train_stop))
        predict_score = list(predict_score)
        result_dict[factor_name + "_predict"] = predict_score
        tcount += 1
    return pd.DataFrame(result_dict)

# 用xgboost模型预测因子IC
def run_xgboost_predict(df, run_type='static', train_start='2008-01-01', train_end='2013-12-31',
                        test_start='2014-01-01', test_end='2017-12-22'):
    '''
    输入参数：
        df: dataframe，包括训练集+测试集的所有数据，columns包括：trade_date(YYYY-mm-dd), 各个因子的真实IC，特征
        run_type: 是静态训练+预测，还是滚动训练+预测
    返回：
        测试期的真实值、预测值的dataframe, columns包括： trade_date, 因子的真实IC，因子的预测IC
    '''
    global fields
    predict_df = pd.DataFrame()  # 列为: trade_date, ROE_predict, EP_predict, ......
    # 静态预测
    if run_type == 'static':
        print("\n====================== run_type:", run_type)
        print("[train]: %s to %s" % (train_start, train_end))
        print("[test]: %s to %s" % (test_start, test_end))
        # 拿到训练、验证、测试集数据
        train_df, val_df, test_df = get_train_val_test_data(df, 0.8, train_start, train_end, test_start, test_end)
        if debug_flag:
            predict_df = predict_factors(train_df, val_df, test_df, ["ROE"])
        else:
            predict_df = predict_factors(train_df, val_df, test_df, fields)
    # 滚动预测
    else:
        print("\n====================== run_type:", run_type)
        predict_df_list = []
        # 根据季度划分时间区间
        date_constrains = get_rolling_date(train_start, test_start, test_end)
        # 在每个时间区间内进行训练和预测
        for [train_s, train_e, test_s, test_e] in date_constrains:
            print("\n---------------- > new rolling periods < ----------------------")
            print("[train]: %s to %s" % (train_s, train_e))
            print("[test]: %s to %s" % (test_s, test_e))
            train_df, val_df, test_df = get_train_val_test_data(df, 0.8, train_s, train_e, test_s, test_e)
            if debug_flag:
                tmp_predict_df = predict_factors(train_df, val_df, test_df, ["ROE"])
            else:
                tmp_predict_df = predict_factors(train_df, val_df, test_df, fields)

            predict_df_list.append(tmp_predict_df)
        # 将不同季度的结果合并到一起
        predict_df = pd.concat(predict_df_list)
    return predict_df


# 通过flag来控制是否要debug
debug_flag = 0


static_time1 = time.time()
# 数据设置
raw_data_dir = "./data_folder/"
fields = ['ROE', 'QuickRatio', 'ETOP', 'TVMA20', 'REVS20', 'negMarketValue']
# 前面得到的数据
model_data = pd.read_pickle(os.path.join(raw_data_dir, 'model_data.pickle'))

real_ic_col = [x for x in model_data.columns if x[-3:] == '_IC']


## 跑static的模型
if debug_flag:
    static_result = run_xgboost_predict(model_data, train_start='2008-01-01', train_end='2008-12-31',
                                       test_start='2014-01-01', test_end='2014-06-30')
else:
    static_result = run_xgboost_predict(model_data,  run_type='static')
    
static_result = static_result.merge(model_data[['trade_date']+real_ic_col], on=['trade_date'], how='left')
static_time2 = time.time()
print("static", static_time2 - static_time1)
print(static_result.head())
static_result.to_pickle(os.path.join(raw_data_dir, "static_result.pickle"))

# 跑rolling的模型
if debug_flag:
    rolling_result = run_xgboost_predict(model_data, run_type='rolling', train_start='2008-01-01',
                                        train_end='2008-12-31', test_start='2014-01-01', test_end='2014-06-30')
else:
    rolling_result = run_xgboost_predict(model_data, run_type='rolling')

rolling_result = rolling_result.merge(model_data[['trade_date']+real_ic_col], on=['trade_date'], how='left')
rolling_time = time.time()
print(rolling_result.head())
rolling_result.to_pickle(os.path.join(raw_data_dir, "rolling_result.pickle"))
print("rolling", rolling_time - static_time2)
```

```python
import os
import time
import numpy as np
import pandas as pd
from quartz.context.parameters import SimulationParameters
from quartz.backtest_tools import get_backtest_data
from quartz.backtest import backtest 
import traceback

############################### 回测行情预加载相关函数和设置 ######################################
'''由于回测数量大，预加载数据可以节省回测时间'''


# 行情预加载函数
def preload_market_service(start, end, universe, benchmark, max_history_window=30, **kwargs):
    parameters = {
        'start': start,
        'end': end,
        'universe': universe,
        'benchmark': benchmark,
        'max_history_window': max_history_window
    }
    sim_parameters = SimulationParameters(**parameters)
    market_service = get_backtest_data(sim_parameters)
    market_service.rolling_load_daily_data(sim_parameters.trading_days)
    return market_service

# 拿到预测出来的IC值
def get_predict_ic(predict_ic_frame, date, factor_name, predict_marker='_predict'):
    '''
    输入参数：
        predict_ic_frame: dataframe，列为 trade_date, [各个因子的真实IC], [各个因子的预测IC]
    返回：
        该因子在当前时间点的预测IC值
    '''
    df = predict_ic_frame.copy()
    if 'trade_date' in df.columns:
        df.set_index('trade_date', inplace=True)
    return df.loc[date, factor_name+predict_marker]
    
# 根据先验知识，判断下期的IC是否有效
def valid_factor(factor_ic, factor_name):
    # 返回因子是否有效，以及因子的方向
    constrains = {"ROE":True, 'QuickRatio':True, "ETOP":True, "TVMA20":False, "REVS20":False, "negMarketValue":False}
    return constrains[factor_name] == (factor_ic >= 0), constrains[factor_name]

# 得到买入的股票list
def get_stock_list(factor_frame, pre_date, factor_name, direction, quantile_num=10):
    '''
    factor_frame: 包含因子值的dataframe，列为： trade_date, secID, [各个因子的名称]
    返回：
    最大或者最小分位的股票列表
    '''
    df = factor_frame[factor_frame.trade_date == pre_date]
    df.dropna(subset=['trade_date', 'secID', factor_name], how='any', inplace=True)
    if len(df) == 0:
        print "Error, no valid factor_frame for date:%s, factor_name:%s" %(pre_date, factor_name)
        return []
    up_boundary = df[factor_name].quantile(1- 1./quantile_num)
    down_boundary = df[factor_name].quantile(1./quantile_num)
    target_df = pd.DataFrame()
    # 正向因子， 取最大组
    if direction: 
        target_df = df[df[factor_name]>=up_boundary]
    else:
        target_df = df[df[factor_name]<=down_boundary]
    return list(target_df.secID.values)
        
# 基于预加载的行情服务进行批量回测的相关设置
def initialize(context):                   # 初始化策略运行环境
    pass

def handle_data(context):                  # 核心策略逻辑
    
    account = context.get_account('fantasy_account')
    pre_date = context.previous_date.strftime("%Y-%m-%d")
    try:
        #################  组合构建 ######################
        global factor_name, input_ic_frame, is_switch, predict_marker
        # print is
        # 拿到IC的预测值
        next_ic = get_predict_ic(input_ic_frame, pre_date, factor_name, predict_marker=predict_marker)
        # 判断因子是否有效
        valid_result, direction = valid_factor(next_ic, factor_name)

        # 如果是不择时，则都买股票
        if not is_switch:
            valid_result=True

        # 决定买入的目标
        if valid_result: # 因子有效，买入因子值最大的分位组（正向因子）/因子值最小的分为组（负向因子）
            target_list = get_stock_list(factor_frame, pre_date, factor_name, direction, quantile_num=10)
        else:
            # print u"%s, 进行择时处理，买入中证500指数"%context.current_date.strftime("%Y-%m-%d")
            target_list = ['000905.ZICN']

        ############### 执行交易 ################
        # 卖出股票
        sell_list = [stk for stk in account.get_positions()]
        for stk in sell_list:
            account.order_to(stk,0)

        # 买入股票
        for stk in target_list:
            account.order_pct_to(stk, 1./len(target_list))
    except:
        err = traceback.format_exc()
        print "Error @ %s in handle_data, skipped" %(context.current_date.strftime("%Y-%m-%d"))
        print "Error info,", err

def get_account():
    accounts = {
                'fantasy_account': AccountConfig(account_type='security', capital_base=10000000)
            }
    return accounts

# 得到并存储回测信息
def get_backtest_result(set_factor_name, ic_predict_frame, set_is_switch=False, set_predict_marker='_predict'):
    '''
    输入参数：
        全局变量： factor_frame, input_ic_frame(列为 trade_date, [因子真实IC], [因子预测IC])
        factor_name： 回测的因子名
        set_is_switch : （是否择时）
    '''
    type_universe = 'ZZ500'
    
    global preload_data, is_switch, input_ic_frame, predict_marker, factor_name
    factor_name = set_factor_name
    is_switch = set_is_switch
    predict_marker = set_predict_marker
    input_ic_frame = ic_predict_frame
    account = get_account()
    
    print u"回测因子 %s, 是否择时：%s" %(factor_name, is_switch)
    backtest_result = pd.DataFrame()
    try:
        # 生成策略对象
        strategy = quartz.TradingStrategy(initialize, handle_data)
        # 开始回测
        sim_params = quartz.SimulationParameters(start, end, benchmark, universe, accounts=accounts, max_history_window=max_history_window,refresh_rate=refresh_rate, commission=Commission(buycost=0, sellcost=0, unit = 'perValue'))
        bt, perf = quartz.quick_backtest(sim_params, strategy, data=preload_data)

        # 回测结果
        tmp = bt[[u'tradeDate',u'portfolio_value',u'benchmark_return']]
        backtest_result.loc[u'年化收益', factor_name] = perf['annualized_return']
        backtest_result.loc[u'年化波动', factor_name] = perf['volatility']
        backtest_result.loc[u'信息比率', factor_name] = perf['information_ratio']
        backtest_result.loc[u'最大回撤', factor_name] = perf['max_drawdown']
    except Exception:
        err = traceback.format_exc()
        print "Error", err
    return backtest_result, bt, perf
```

## 4.1 合成因子
- 合成方法：

  正向因子，下一期预测的IC为正，因子权重为预测的IC值
  
  负向因子，下一期预测的IC为负，因子权重为预测的-IC值
  
  所有因子都相反，则等权合成
  
  最后进行归一化
  
  等权合成因子作为对比

```python
# 根据因子方向调整因子，并得到因子权重
# 正向因子，IC>0, 则为IC本身，IC<0, 则为0（负向因子类似）
def adjust_weight_direction(input_frame, factor_lists, predict_marker='_IC_extend'):
    '''
    输入参数：
        input_frame: dataframe, columns为: trade_date, 因子值列 (输入的因子值列可以有真实值，或者预测值)
    输出：
        adjusted_frame: dataframe, trade_date为index, columns为: 调整之后的因子值列 （输出的因子值列，只有包含predict_marker的列）
    '''
    constrains = {"ROE":True, 'QuickRatio':True, "ETOP":True, "TVMA20":False, "REVS20":False, "negMarketValue":False}
    df = input_frame[['trade_date']+[x for x in input_frame.columns if predict_marker in x]]
    df.index = range(len(df))
    for factor_name in factor_lists:
        if not constrains[factor_name]:
            multiplyer = -1
        else:
            multiplyer = 1
        df[factor_name+predict_marker] *= multiplyer
    df.set_index("trade_date", inplace=True)
    # 调整一下列的顺序, 和factor_lists保持一致
    df = df[[x+predict_marker for x in factor_lists]]
    df[df<0] = 0
    df['total'] = df.sum(axis=1)
    target_list = [1]*len(factor_lists)
    target_list.extend([sum(target_list)])
    df.loc[df[df.total==0].index.values, :] = target_list
    # 权重归一化
    for factor_name in factor_lists:
        df[factor_name+predict_marker] /= df['total']
    del df['total']
    return df


# 根据因子的权重，调整因子值，并得到合成因子值，命名为combined_col_name
def adjusted_factor_value(ori_factor_frame, adjusted_weight_frame, factor_lists, predict_marker, combined_col_name):
    '''
    ori_factor_frame: 原来的因子值dataframe，列为:trade_date, secID, [各个因子列]
    adjusted_weight_frame: 因子权重列，index为 trade_date, columns为【各个因子的权重，列名是因子名+predict_marker】
    combined_col_name：合成后的因子值列
    '''
    constrains = {"ROE":1, 'QuickRatio':1, "ETOP":1, "TVMA20":-1, "REVS20":-1, "negMarketValue":-1}
    weight_frame = adjusted_weight_frame.reset_index()
    combined_frame = ori_factor_frame.merge(weight_frame, on=['trade_date'], how='inner')
    for factor_name in factor_lists:
        combined_frame['combined_'+factor_name] = constrains[factor_name]*combined_frame[factor_name] * combined_frame[factor_name+predict_marker]
    combined_frame[combined_col_name] = combined_frame[['combined_'+x for x in factor_lists]].sum(axis=1)
    return combined_frame[['trade_date', 'secID', combined_col_name]]

# 风格动量组IC权重
adjusted_weight_baseline = adjust_weight_direction(baseline_frame, fields, predict_marker='_IC_extend')

# 静态xgboost模型IC权重
adjusted_weight_static = adjust_weight_direction(static_result, fields, predict_marker='_predict')

# rolling 的 xgboost模型IC权重
adjusted_weight_rolling = adjust_weight_direction(rolling_result, fields, predict_marker='_predict')


# 根据权重合成因子
static_factor_frame = adjusted_factor_value(factor_frame, adjusted_weight_static, fields, '_predict', 'static')

rolling_factor_frame = adjusted_factor_value(factor_frame, adjusted_weight_rolling, fields, '_predict', 'rolling')

# 等权组
equal_factor_frame = factor_frame.copy()
equal_factor_frame['weight_equal'] = equal_factor_frame[fields].mean(axis=1)
equal_factor_frame = equal_factor_frame[['trade_date', 'secID', 'weight_equal']]
equal_factor_frame = equal_factor_frame[(equal_factor_frame.trade_date >='2014-01-01') & (equal_factor_frame.trade_date<='2017-12-22')]

# 将三组合成到一个dataframe，便于回测
total_factor_frame = equal_factor_frame.merge(static_factor_frame, on=['trade_date', 'secID'], how='inner')
total_factor_frame = total_factor_frame.merge(rolling_factor_frame, on=['trade_date', 'secID'], how='inner')
total_factor_frame.head()
```

```python
# 某日的ST股票股票
def get_st_stocks(source_universe, st_date=None):
    """
    Args:
        source_universe (list of str): 需要进行筛选的股票列表
        st_date (datetime): 进行筛选的日期,默认为调用当天
    Returns:
        list: ST股票列表
    """
    st_date = st_date if st_date is not None else datetime.datetime.now().strftime('%Y%m%d')
    df_ST = DataAPI.SecSTGet(secID=source_universe, beginDate=st_date, endDate=st_date, field=['secID'])
    return list(df_ST['secID'].values)

def hedge_return_drawback(bt, perf, title):
    """
    对冲之后的组合净值以及回撤
    输入：
        bt：quartz回测结束自动生成的dict
        title：str
    返回：
        ax：matplotlib figure 对象
    """
    bt_quantile_ten = bt
    data = bt_quantile_ten[[u'tradeDate',u'portfolio_value',u'benchmark_return']]
    data['portfolio_return'] = data.portfolio_value/data.portfolio_value.shift(1) - 1.0
    data['portfolio_return'].ix[0] = data['portfolio_value'].ix[0]/	10000000.0 - 1.0
    # 超额收益率
    data['excess_return'] = data.portfolio_return - data.benchmark_return
    data['excess'] = data.excess_return + 1.0
    # 累计超额收益率
    data['excess'] = data.excess.cumprod()

    df_cum_rets = data['excess']
    running_max = np.maximum.accumulate(df_cum_rets)
    # 回撤值
    underwater = -((running_max - df_cum_rets) / running_max)
    underwater.index = data['tradeDate']
    
    # 画图显示对冲净值以及回撤
    fig = plt.figure(figsize=(12, 5))
    fig.set_tight_layout(True)
    ax1 = fig.add_subplot(111)
    ax2 = ax1.twinx()
    x = range(len(underwater))
    ax2.grid(False)
    # ax1.set_ylim(-0.30, 0)
    ax1.set_ylabel(u'回撤', fontproperties=font, fontsize=16)
    ax1.fill_between(underwater.index, 0, np.array(underwater), color='#000066', alpha=1)
    ax2.set_ylabel(u'净值', fontproperties=font, fontsize=16)
    ax2.plot(data['tradeDate'], data[['excess']], label='hedged(right)', color='r')
    # ax2.set_ylim(bottom=0.9, top=7)
    s = ax1.set_title(title, fontproperties=font, fontsize=16)
    statistic_result = excess_statistic(data, perf, underwater)
    return fig, statistic_result

# 统计对冲后的指标
def excess_statistic(data, perf, underwater):
    # 年化收益率
    annual_return = (data['excess'].values[-1] - 1)*250/len(data)
    # 年化波动率
    annual_std = pow(250, 0.5)*data['excess_return'].std()
    # 信息比
    ir = annual_return/annual_std
    # 最大回撤
    max_drawback = -1*min(underwater)
    # 累计收益率
    accum_ret = data['excess'].values[-1]-1
    # 返回
    return_df = pd.DataFrame()
    return_df.loc[u'年化收益', 'group'] = annual_return
    return_df.loc[u'年化波动率', 'group'] = annual_std
    return_df.loc[u'信息比', 'group'] = ir
    return_df.loc[u'最大回撤', 'group'] = max_drawback
    return_df.loc[u'累计收益', 'group'] = accum_ret
    return return_df.round(4)
```
