# 4.CTA策略Aberration在股市中的应用

#### 所谓CTA，即商品交易顾问(Commodity Trading Advisor)，一般指期货行业中的一种基金管理方式。

#### CTA策略一般分为两种：趋势跟随型策略（动量）和套利策略（均值回复）：
#### 趋势跟随型策略利用大量的模型寻找当前的市场趋势，判断多空，只要存在大幅度的上涨和下跌，则都能获利，但是在市场波动较小或震荡市期间，这种策略可能因为不停地止损而出现回撤。
#### 套利策略则主要应用在跨期、跨品种配对交易中，根据价差反转套利，和股票市场中的高抛低吸类似，核心思想是利用历史走势基本一致地对资产进行套利。

#### 常见的CTA策略有：Dual Thrust策略、R-Breaker策略、ATR突破策略、长线Aberration策略、海龟交易策略。

#### 本次研究基于股票市场，相比于期货市场有两大限制：
#### 1.不能进行日内交易
#### 2.不能做空
#### 所以我们的策略是日间交易的只做多的策略，考虑到要避免频繁地开仓，我们选择长线Aberration作为较适合股票市场的CTA策略。

### 长线Aberration策略

#### Aberration策略由Keith Fitschen于1986年发明。在1993年，他将该系统发布在美国《Future Truth》杂志上，自策略发布之日起，其绩效一直排名靠前，在1997年、2001年和2005年已发布交易系统的业绩排名中均名列第一。
#### 该策略的思想很简单，就是利用移动平均线和标准差设置通道，然后随时判断是否突破了通道，从而捕捉到趋势,当股价远离基准价格超过一定幅度则可能形成长期趋势。而短期股价偏离基准价格过多则可能因过度偏离均值而形成大幅回撤甚至趋势终止。故我们需要开仓价、止盈价、止损价。
- 开仓价：当日最高价 > 均价 + 开仓触发倍数 × 观察期内标准差最大值----股价突破了通道
- 止盈价：当天最高价 > 均价 + 止盈触发倍数 × 观察期内标准差最大值----短期股价偏离基准价格过多可能出现大幅回撤
- 止损价：当天最低价 < max(均价， 开仓价 - 止损触发倍数 × 开仓时观察期内标准差最大值)----固定止损价与移动止损价的结合

#### 本文的研究思路按照如下的四个步骤进行：
#### 1.数据准备		
#### 2.策略思想
#### 3.计算策略收益
#### 4.基础回测指标构建

一.数据准备
通过DataAPI.MktEqudGet获取个股数据
求出均价、标准差，进而计算出开仓价、止盈价
通过np.where函数确定开仓信号和止盈信号

```python
import pandas as pd
import numpy as np

length = 10            
# 观察周期长度
open_trigger = 0.5      
# 价格向上偏离均线0.5倍观察期内标准差的最大值开仓
stopwin_trigger = 3     
# 价格向上偏离均线3倍观察期内标准差的最大值止盈
stoplose_trigger = 1    
# 移动止损：开仓价向下偏离观察期内标准差的最大值止损

data = DataAPI.MktEqudAdjGet(secID=u"",ticker=u"600030",tradeDate=u"",beginDate=u"20130101",endDate=u"20180101",isOpen="1",field=u"",pandas="1")
# 获取个股数据
data['return'] = data['closePrice'].pct_change()
# 每日收益率
data['ma'] = data['closePrice'].rolling(window=length).mean()
# 观察期内均线
data['std'] = data['closePrice'].rolling(window=length).std()
# 观察期内标准差
data['std_limit'] = data['std'].rolling(window=length).max()      
# 观察期内标准差的最大值，用最大值而非当日值可减少开平仓的频率

data['ma_s1'] = data['ma'].shift(1)
data['std_limit_s1'] = data['std_limit'].shift(1)            
data['long_open_price'] = data['ma_s1'] + data['std_limit_s1'] * open_trigger    
# 计算每一天满足条件的开仓价,价格高于0.5倍观察期内标准差最大值开仓
data['long_stopwin_price'] = data['ma_s1'] + data['std_limit_s1'] * stopwin_trigger   
# 计算每一天满足条件的止盈价,价格高于3倍观察期内标准差最大值止盈

data['long_open_signal'] = np.where(data['highestPrice'] > data['long_open_price'], 1, 0)
# 当最高价 > 开仓价时，开仓信号为1，否则为0
data['long_stopwin_signal'] = np.where(data['highestPrice'] > data['long_stopwin_price'], 1, 0)
# 当最高价 > 止盈价时，平仓信号为1，否则为0
```

二.策略思想
在持仓和空仓两种情形下分别判断，计算策略当天收益率

```python
flag = 0    
# 记录持仓情况，0代表空仓，1代表持仓；

for i in range(12, (len(data)-1)):
        # range(12, (len(data)-1))为有效的数据时间段
        if flag == 1:           
            stoplose_price = max(data.loc[i,'ma_s1'], open_price - data.loc[i, 'std_limit_s1'] * stoplose_trigger)
            # 计算止损价格，取均线和开仓价下移一定倍数标准差，两者的最大值作为止损价
            
            if data.loc[i, 'long_stopwin_signal']: 
                # 当long_stopwin_signal为1时，判断为True，触发止盈条件
                data.loc[i, 'strategy_return'] = data.loc[i, 'long_stopwin_price'] / data.loc[i-1, 'closePrice'] - 1
                flag = 0
                # 多头止盈并计算当日收益率    
                        
            elif data.loc[i, 'lowestPrice'] < stoplose_price:  
                # 当最低价 < 止损价时，触发止损条件          
                data.loc[i, 'strategy_return'] = min(data.loc[i, 'openPrice'], stoplose_price) / data.loc[i-1, 'closePrice'] - 1
                flag = 0
                # 考虑到当天开盘价就低于止损价，无法止损的情况，取止损价和开盘价的最小值
                # 多头移动止损并计算当日收益率          
            
            else: 
                data.loc[i, 'strategy_return'] = data.loc[i, 'closePrice'] / data.loc[i-1, 'closePrice'] - 1
                # 多头持仓并计算收益率
        
        else:
        # 无持仓进行开仓
            if data.loc[i, 'long_open_signal']:    
                # 当long_open_signal为1时，判断为True，触发开仓条件
                flag = 1

                open_price = max(data.loc[i, 'openPrice'], data.loc[i, 'long_open_price'])    
                # 需要比较当天的开盘价和开仓价，当开盘价高于开仓价时，只能以开盘价进行开仓，不能用开仓价
                # 否则会导致策略收益高估
                              
                data.loc[i, 'strategy_return'] = data.loc[i, 'closePrice'] / open_price - 1
                # 记录当天盈利情况  
                
                stoplose_price = max(data.loc[i,'ma_s1'], open_price - data.loc[i, 'std_limit_s1'] * stoplose_trigger)
                # 计算止损价格，取均线和开仓价下移一定倍数标准差，两者的最大值作为止损价
                                     
                if (data.loc[i, 'lowestPrice'] < stoplose_price or data.loc[i, 'long_stopwin_signal']):    
                    # 若开仓当天同时满足止损条件或止盈条件                   
                        data.loc[i, 'strategy_return'] = data.loc[i + 1, 'openPrice'] / open_price - 1    
                        flag = 0
                        # 因为是T+1交易，不能当天平仓，所以以第二天开盘价平仓                       
```

三.计算策略收益
（收益率 + 1）.cumprod方法计算离散的累积收益率

```python
data['strategy_return'].fillna(0, inplace=True)
data['strategy_return_cum'] = (data['strategy_return'] + 1).cumprod()
data['return_cum'] = (data['return'] + 1).cumprod()
#计算策略和benchmark的累积收益率

data[['return_cum', 'strategy_return_cum']].plot()
#画出累积收益率图
```

四.基础回测指标构建

构建一个简单的ratio函数来判断收益率、波动率、最大回撤三个指标。

这里计算收益率的方式和以往data.mean() * 252的方式不同，是Return ** (252 / N) - 1的思想，其中Return为累积收益率，N为回测交易日数量。
计算的是复利下的年化收益率，相比之前的方式更好。

最大回撤的计算方法也做了改进，之前是回撤的收益率的倍数，现在是回撤的百分比。

注意：输入的数据为每日收益数据，非累积收益数据。

```python
def ratio(data):
    strategy_cum = (data + 1).cumprod()
    return_year = strategy_cum.tolist()[-1] ** (252./strategy_cum.shape[0]) - 1    
    #计算收益率
    
    volatility = data.std() * 252 ** 0.5
    #计算波动率
    
    drawdown = ((strategy_cum.cummax() - strategy_cum)/strategy_cum.cummax()).max()
    #计算最大回撤
     
    res = pd.DataFrame([return_year, volatility, drawdown], columns=[data.name], index=['return', 'volatility', 'drawdown'])
    #输出数据为列表[收益率，波动率，最大回撤]， 列名为[data.name]，是取Series下列名的方法， 索引为自定义的['return', 'volatility', 'drawdown']
    return res

ratio(data['return']).join(ratio(data['strategy_return']))
#DataFrame下的join方法，按index合并，合并benchmark和策略的回测指标
```

## 总结：
- 首先从收益曲线上可以看出，长线Aberration策略在震荡型市场中的表现很一般，但在趋势形成时效果非常好，说明它确实是一种不错的趋势型CTA策略。
- 其次相比于benchmark，策略在收益率、波动率、最大回撤方面均大幅优于benchmark，尤其是最大回撤一项有巨大改进。

- 不足之处：
  没有考虑交易费用和滑点。
  当一天内价格同时触发开仓、止盈、止损条件时，根据不同的情形应有更好的处理方式。这里只做了第二天平仓的近似处理。
  股票市场的T+1制度和做空限制，使得CTA策略不能发挥百分百的功能。
