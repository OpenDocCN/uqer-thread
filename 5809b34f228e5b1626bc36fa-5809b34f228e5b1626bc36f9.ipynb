```python
import pandas as pd
import numpy as np
from datetime import datetime
from datetime import timedelta
from sklearn.svm import SVC
from CAL.PyCAL import *

start = '2014-01-01'                       # 回测起始时间
end='2016-10-18'                       # 回测结束时间
benchmark = 'HS300'                        # 策略参考标准
universe=DynamicUniverse('HS300')+DynamicUniverse('ZZ500')
capital_base = 1000000                      # 起始资金
freq = 'd'                                 # 策略类型，'d'表示日间策略使用日线回测，'m'表示日内策略使用分钟线回测
refresh_rate = (1, ['09:40', '14:50'])  # 表示每5个交易日，在那天 09:40 和 14:50 时候调仓 
max_history_window = 120  ## 设置的最大历史行情窗口5小于调用的历史行情窗口20


used_factors = ['EMA10', 'EMA60', 'ROA', 'PE', 'LCAP', 'DHILO', 'DebtEquityRatio', 'OperatingProfitGrowRate', 'TotalAssetGrowRate', 'NPToTOR']
index_factors=['secID','tradeDate']

def initialize(account):                   # 初始化虚拟账户状态
    pass

def handle_data(account):                  # 每个交易日的买入卖出指令
    
    #时间节点，股票池
    timepoint=account.current_date
    train_start=timepoint-timedelta(525)
    train_end=timepoint-timedelta(5)
    train_start=train_start.strftime("%Y-%m-%d")
    train_end=train_end.strftime("%Y-%m-%d")
    yesterday=account.previous_date
    
    #股票池合规
    universe=account.current_universe
    # 2. 去除ST股
    try:
        STlist = DataAPI.SecSTGet(secID=universe, beginDate=yesterday, endDate=yesterday, field=['secID']).tolist()
        universe = [s for s in universe if s not in STlist]
    except:
        pass

    # 3. 去除流动性差的股票
    tv = account.get_attribute_history('turnoverValue', 20)
    mtv = {sec: sum(tvs)/20. for sec, tvs in tv.iteritems()}
    universe = [s for s in universe if mtv.get(s, 0) >= 10**7]

    # 4. 去除新上市或复牌的股票
    opn = account.get_attribute_history('openPrice', 1)
    universe = [s for s in universe if not (np.isnan(opn.get(s, 0)[0]) or opn.get(s, 0)[0] == 0)]
    
    #止损
    own_list=account.security_position
    own_list=pd.Series(own_list)
    own_list=own_list.index.values
    own_list=list(own_list)
    #时间节点，股票池
    timepoint=account.current_date
    stop_start=timepoint-timedelta(3)
    stop_end=timepoint-timedelta(1)
    tmp_price=DataAPI.MktEqudGet(tradeDate=stop_end,secID=own_list,isOpen="",field=u"secID,tradeDate,closePrice",pandas="1")
    #昨日
    tmp_t_price=DataAPI.MktEqudGet(tradeDate=stop_start,secID=own_list,isOpen="",field=u"secID,tradeDate,closePrice",pandas="1")
    #更早
    tmp_price['updown']=(tmp_price['closePrice']-tmp_t_price['closePrice'])/tmp_price['closePrice']
    for i in tmp_price.index:
        if tmp_price.ix[i,['updown']].values>=-0.065:
            tmp_price.ix[i,['updown']]=np.nan
    tmp_price=tmp_price.dropna()
    tmp_price=tmp_price.set_index('secID')
    stop_list=tmp_price.index.values
    stop_list=list(stop_list)
    if len(stop_list)>0:
        print account.current_date.strftime("%Y-%m-%d"), ' 超跌股票代码 。'
        print stop_list
    for stk in stop_list:
        try:
            order_to(stk, 0)
        except:
            print 'wrong with ',stock,' cannot sell'
    #剔除超跌股票
    universe = [s for s in universe if s not in stop_list]    
        
    #获取训练集，使用半年数据回归，预测10天涨跌
    Xtrain,Ytrain=train_data(universe = universe , used_factors = used_factors , start = train_start , end = train_end , ad_period = 10)
    print account.current_date, '  Data acquisition has been completed . The timepoint is ', datetime.now()
    
    #SVM训练
    i=10
    g=1.0
    clf=SVC(C=i,cache_size=2100,gamma=g,decision_function_shape='ovo')
    clf.fit(Xtrain,Ytrain)
    
    #预测数据
    buylist=[]
    idate=Date(account.previous_date.year,account.previous_date.month,account.previous_date.day).strftime('%Y%m%d')
    tmp=DataAPI.MktStockFactorsOneDayGet(tradeDate=account.previous_date,secID=universe,field=index_factors+used_factors,pandas="1").dropna().set_index('secID')
    for ifac in used_factors:
        tmp[ifac]=pd.Series(standardize(neutralize(winsorize(tmp[ifac].to_dict(),n_draw=5),idate)))  
    for istock in universe:
        try:
            xtmp=tmp.loc[istock,:]
            xtmp=xtmp.iloc[1:]
            xtmp=np.mat(xtmp)
            ytmp=clf.predict(xtmp)
            if ytmp>=1:
                buylist.append(istock)
        except:
            #print 'wrong with stock ',istock
            pass
    
    rebalance(account, buylist)
    return

def rebalance(account, buylist):
    for stock in account.security_position:
        if stock not in buylist:
            try:
                order_to(stock, 0)
            except:
                print 'wrong with ',stock,' cannot sell'
    if len(buylist)<=10:
        length=10
    else:
        length=len(buylist)
    for stock in buylist:
        try:
            order(stock, account.reference_portfolio_value / (1.67*length) / account.reference_price[stock])
        except:
            pass
    return        
        
def train_data(universe,used_factors,start = '2014-05-01',end = '2014-10-30',ad_period=10):

    cal = Calendar('China.SSE')
    date=cal.bizDatesList(start,end)
    
    set_factors=['tradeDate','secID']
    
    training_data_set=pd.DataFrame()
    
    
    for i,idate in enumerate(date):
        if i+ad_period>=len(date):
            continue
        if i%10!=0 :
            continue 
        else:
            t=i+ad_period
            tdate=date[t]
            tmp_price=DataAPI.MktEqudGet(tradeDate=idate,secID=universe,isOpen="",field=u"secID,tradeDate,closePrice",pandas="1")
            tmp_t_price=DataAPI.MktEqudGet(tradeDate=tdate,secID=universe,isOpen="",field=u"secID,tradeDate,closePrice",pandas="1")
            tmp_price['updown']=(tmp_t_price['closePrice']-tmp_price['closePrice'])/tmp_price['closePrice']
            # 设置涨跌幅度分级
            for i in tmp_price.index:
                if tmp_price.ix[i,['updown']].values>=0.08:
                    tmp_price.ix[i,['updown']]=2
                elif tmp_price.ix[i,['updown']].values>=0.03 and tmp_price.ix[i,['updown']].values<=0.08:
                    tmp_price.ix[i,['updown']]=1
                elif tmp_price.ix[i,['updown']].values<=0.03 and tmp_price.ix[i,['updown']].values>=-0.03:
                    tmp_price.ix[i,['updown']]=0
                elif tmp_price.ix[i,['updown']].values<=-0.03 and tmp_price.ix[i,['updown']].values>=-0.08:
                    tmp_price.ix[i,['updown']]=-1
                elif tmp_price.ix[i,['updown']].values<=-0.08:
                    tmp_price.ix[i,['updown']]=-2
            tmp_price['updown']=tmp_price['updown'].astype(int)

            #因子参数
            tmp=DataAPI.MktStockFactorsOneDayGet(tradeDate=idate,secID=universe,field=set_factors+used_factors,pandas="1")
            #数据合并，因子处理
            tmp=pd.merge(tmp,tmp_price)
            tmp=tmp.set_index(['secID'])
            tmp=tmp.dropna()
            for ifac in used_factors:
                tmp[ifac]=pd.Series(standardize(neutralize(winsorize(tmp[ifac].to_dict(),n_draw=5),idate.strftime('%Y%m%d'))))  
            training_data_set=pd.concat([training_data_set,tmp])
    #重新设置index
    rownum=range(len(training_data_set))
    rownum=pd.Series(rownum)
    training_data_set=training_data_set.set_index(rownum)  

    Xtrain=training_data_set.loc[:,used_factors]
    Xtrain=np.mat(Xtrain)
    Ytrain=training_data_set.loc[:,'updown']
    Ytrain=np.mat(Ytrain).T
    return Xtrain,Ytrain    
```
