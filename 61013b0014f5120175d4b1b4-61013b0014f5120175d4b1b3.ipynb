# Pandas  Concat

```python
#pd.concat( objs, axis=0, join='outer', join_axes=None, ignore_index=False, keys=None, levels=None, names=None, verify_integrity=False, sort=None, copy=True,)
#一：默认的是按照行来合并  二：默认的是外连接  三：keys代表保留原来的信息  四：names拼接后形成的数据结构添加名字
import pandas as pd
import numpy as np
df1=pd.DataFrame(np.ones((3,4))*0,columns=['a','b','c','d'],index=[1,2,3])
df2=pd.DataFrame(np.ones((3,4))*1,columns=['b','c','d','e'],index=[2,3,4])
print(df1)
print(df2)
res=pd.concat([df1,df2],axis=0)#concat 默认的就是按照行来合并,两者都有的部分全部呈列，只存在其中一个的用nan表示
print(res)
res=pd.concat([df1,df2],join='outer')#concat 默认的是外连接，即join=‘outer’
print(res)
ref=pd.concat([df1,df2],axis=1,keys=['df1','df2'],names=['我最屌'])
print(ref)
print('inner:')
res=pd.concat([df1,df2],join='inner',ignore_index=True)
print(res)
```
