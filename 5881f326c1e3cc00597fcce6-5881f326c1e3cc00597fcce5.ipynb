# 数据结构-series

- series长啥样？series有点像dictionary字典，简单说就是**索引+数值**

```python
d = {'W':1,'D':2,'T':3}#字典
print d
s = pd.Series(d)
print s
```

- series怎么计算的？series赋值计算同numpy，这里造两个series玩一玩，注意：**空缺位会被NaN（not a number）取代**

```python
s1 = pd.Series({'a':1,'b':2,'c':3})
s2 = pd.Series({'b':4,'c':5,'d':6})
s3 = s1+s2
print s3
```

- series有些什么特点？一个特点是计算过程中索引会自动对齐，上面那个例子已经说明这个特质；还有一个有趣的特点是**索引和数值均有name属性**,下面分别对索引及数值命名

```python
s1.index.name = 'index'
s1.name = 'values'
s1
```

PS:命名有什么好处？当你要处理维护多张数据表格的时候，通过索引名或数值名可以**快速定位数据**，同时打印出来后**方便阅读**。

# series的小心思

- 卧槽，每次操作都有NAN怎么玩？处理数据时，nan非常讨厌，无论删除赋值都需要话费心思，这里的isnull函数能够**返回series中nan的bool值**，对筛选nan很有帮助

```python
s3.isnull()
```

# 数据结构-dataframe

dataframe是pandas中一直会用到的，处理的就是这玩意儿。相比series，dataframe多了一列columns,这样看上去有点像一个二维数组了。

```python
dt = pd.DataFrame(s3,columns=['A'])
print dt
```

- 怎么造一个dataframe？这里有4种比较常用的方法

```python
dt1 = pd.DataFrame(index=['a','b','c','d'],columns=['A','B','C','D'],data={'A':1,'B':2,'C':3,'D':4})#最经典的传统做法
print dt1
```

```python
dt2 = pd.DataFrame({'A':s1,'B':s2})#series的字典
print dt2
```

```python
dt3 = pd.DataFrame({'A':{'a':1,'b':2},'B':{'c':3}})#字典的字典
print dt3
```

```python
import numpy as np
dt4 = pd.DataFrame(np.ones((3,2)))#二维numpy数组
print dt4
```

- dataframe较series属性特点上有哪些差异？一般的计算规则属性是一样的，唯一**不同点在于取值有了轴方向**，这里的轴方向即行方向和列方向，下面介绍三种常用操作方式

```python
print dt1['A']#取dataframe的列，注意返回值变成了series
```

```python
print dt2.ix['a']#取dataframe的行，有兴趣的朋友可以百度iloc，loc，ix的区别
```

```python
del dt3['B']#删除dataframe的列
print dt3
```

- dataframe还有哪些有趣的特性？一个values属性和索引不可修改属性，这个索引index比较高冷，不能随便搞

```python
dt1.values#返回了一串2维ndarray
```

下面着重讲讲常用的dataframe搞法……

```python
dt1.index.diff(dt2.index)#diff取两个series的差
```

```python
dt1.index.intersection(dt3.index)#两个series的交集
```
