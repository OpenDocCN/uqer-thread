```python
import pandas as pd
import numpy as np
beginDate="2010-01-01"
endDate="2018-02-12"
fields= ['tradeDate','secID','openPrice','closePrice']
secID= ['600000.XSHG','300104.XSHE','600519.XSHG'] #设置样本股票
df= DataAPI.MktEqudAdjGet(secID=secID,beginDate=beginDate,endDate=endDate,field=fields,isOpen='1',pandas="1") 
df1= df[df.secID=='300104.XSHE'][-5:-1] #剔除最后一个交易日
df2= df[df.secID=='600000.XSHG'][-10:]
df3= df[df.secID=='600519.XSHG'][-4:] #取最后4个交易日
df1=df1.append(df2)
df1=df1.append(df3) #组成测试数据
print '1) ------------------------------------------'
print '包含不规则的全部样本数据，如下：'
print df1 
#以上代码模拟取到一组不规整的数据，即不等长（或因前后有停牌）

df1.set_index(['secID','tradeDate'],inplace=True) #对代码和日期索引
print '看到以代码和日期排序了，方便groupby切片操作'
print df1 
df1.reset_index(inplace=True) #撤销排序列表，但保持了排序后的队形
print
print '2) ------------------------------------------'
print '取到了所有排好序的数据（关键是日期），没有列索引：'
print df1
print 
print '3) ------------------------------------------'
df_mean = pd.DataFrame([], columns=['secID','openMean','closeMean']) #创建空表，保存满足条件的移动均值
print '用迭代方法遍历所有股票，并切片最后6个交易日：'
for key_secID, sub_df_asKey in df1.groupby(['secID'])['tradeDate','openPrice','closePrice']:
    print key_secID
    print sub_df_asKey[-6:] #对每个分组切endDate的前6个交易日
    lastDate= sub_df_asKey.iloc[-1:].tradeDate.values[0] #取最后一个交易日的日期，并转为数据型
    if lastDate == endDate: #剔除
        print sub_df_asKey[-6:].mean() #求出最后6个交易日的平均值
        listRow=  [key_secID] + sub_df_asKey[-6:].mean().values.tolist() #组成list数据
        print listRow
        df_mean=df_mean.append(pd.DataFrame([listRow],columns=['secID','openMean','closeMean']),ignore_index=True) #增加行数据
    #end if
#end for
print
print '4) ------------------------------------------'
print '最后得到分组后有交易日的最后6个交易日的平均值，剔除了300104,而600519只有4个交易日'
print df_mean
```
