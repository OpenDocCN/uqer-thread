1. 思路来源于华泰联合 2011年9月的多因子选股策略（数量化选股策略之十二）
2. 基本步骤：
（1）选取十个因子（前10个报告分析得出每个系列表现最好的因子），等权重——总市值，市盈率TTM，营业利润同比增长率、净资产收益率、前1个月涨跌幅、前1个月日均换手率、前1个月波动率、户均持股比例变化、机构持股比例变化、近1个月净利润上调幅度（最新当年净利润预测平均值/1个月前当年净利润预测平均值-1）
（2）对各因子进行标准化处理：：（因子值—因子均值）/因子标准差(有函数standardize（）)
（3）综合得分=sum[该只股票各正向因子（营业利润同比增长率、净资产收益率ROE、户均持股比例变化、机构持股变化、最近1个月净利润上调幅度）标准化之后的值]-sum[各负向因子（总市值LCAP、市盈率TTMPE、前1个月涨跌幅、前1个月日均换手率VOL20、前1个月波动率）标准化之后的值]
(4)将综合得分由高到低排序sort(ascending=False)
(5)取前20个进入账户
