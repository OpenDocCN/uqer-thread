# 1. 概述
---
我们发现成交量泄露了知情交易者的行动，换手率与第二天股价的波动呈现出正相关。
- 这种现象暗示，股票大跌前知情交易者大量抛售股票急于离场，而大涨前知情交易者大量购买股票急于进场。
- 换手率和股价的波动关系越强，则暗示信息泄露 程度越高，我们在该类股票上的博弈可能陷入劣势。 

# 2. 数据处理
---
用涨跌幅的绝对值来代表市场消息的冲击，突如其来的好消息和 坏消息都会导致股票涨跌幅绝对值扩大。
我们考察 **T 日标准化换手率 与 T+1 日标准化涨跌幅的绝对值的关系**

 
 （1） 换手率定义为“成交额/自由流通市值” 
 （2） 剔除个股涨跌停、停牌、交易量过低等异常交易数据 
 （3） 标准化换手率的定义为当日换手率减去过去 20 个交易日 的平均换手率，𝑆𝑉𝑡 = 𝑉𝑡 − 1 /20 ∗∑ (𝑉𝑖)𝑡−20 𝑡−1 
 （4） 标准化涨跌幅的定义为当日涨跌幅减去过去 20 个交易日 的平均涨跌幅，𝑆𝑅𝑡 = 𝑅𝑡 − 1 /20 ∗∑ (𝑅𝑖)𝑡−20 𝑡−1 

```python
import matplotlib.pyplot as plt
from matplotlib import rc
from matplotlib import dates
rc('mathtext', default='regular')
import seaborn as sns
sns.set_style('darkgrid')
import datetime
import numpy as np
import pandas as pd
import time
import scipy.stats as st
from CAL.PyCAL import * 

cal = Calendar('China.SSE')
```

```python
begin_date = '20100101'  
end_date = '20180101'    
universe = set_universe('HS300')
```

```python
def getTurnoverRateAll(universe, begin, end):
    
    ret_data = pd.DataFrame()
    for stk in universe:
        data = DataAPI.MktEqudAdjGet(secID=stk, beginDate=begin, endDate=end, 
                                     field='secID,tradeDate,turnoverRate')    
        tmp_ret_data = data.sort('tradeDate')
        tmp_ret_data = tmp_ret_data[['tradeDate','turnoverRate']]
        tmp_ret_data.columns = ['tradeDate', stk]

        if ret_data.empty:
            ret_data = tmp_ret_data
        else:
            ret_data = ret_data.merge(tmp_ret_data, on='tradeDate', how='outer')

    return ret_data


def getForwardReturnsAll(universe, begin, end):

    ret_data = pd.DataFrame()
    for stk in universe:
        data = DataAPI.MktEqudAdjGet(secID=stk, beginDate=begin, endDate=end, 
                                     field='secID,tradeDate,closePrice')    
        tmp_ret_data = data.sort('tradeDate')
        
        tmp_ret_data['forwardReturns'] = tmp_ret_data['closePrice'].shift(-1) / tmp_ret_data['closePrice']  - 1.0
        tmp_ret_data = tmp_ret_data[['tradeDate','forwardReturns']]
        tmp_ret_data.columns = ['tradeDate', stk]

        if ret_data.empty:
            ret_data = tmp_ret_data
        else:
            ret_data = ret_data.merge(tmp_ret_data, on='tradeDate', how='outer')

    return ret_data


def excess_returns_plot(fac, periods, title):
    n_quantile = 5
    cols_mean = [i+1 for i in range(n_quantile)]
    cols = cols_mean

    excess_returns_means = pd.DataFrame(index=fac.index, columns=cols)

    for dt in excess_returns_means.index:
        qt_mean_results = []

        tmp_illiq = fac.ix[dt].dropna()
        tmp_return = periods.ix[dt].dropna()
        tmp_return = tmp_return[tmp_return<0.6]
        tmp_return_mean = tmp_return.mean()

        pct_quantiles = 1.0/n_quantile
        for i in range(n_quantile):
            down = tmp_illiq.quantile(pct_quantiles*i)
            up = tmp_illiq.quantile(pct_quantiles*(i+1))
            i_quantile_index = tmp_illiq[(tmp_illiq<=up) & (tmp_illiq>=down)].index
            mean_tmp = tmp_return[i_quantile_index].mean() - tmp_return_mean
            qt_mean_results.append(mean_tmp)

        excess_returns_means.ix[dt] = qt_mean_results

    excess_returns_means.dropna(inplace=True)

    # 分组的超额收益作图
    fig = plt.figure(figsize=(10, 6))
    ax1 = fig.add_subplot(111)

    excess_returns_means_dist = excess_returns_means.mean()
    lns2 = ax1.bar(excess_returns_means_dist.index, excess_returns_means_dist.values, align='center', alpha=0.5, width=0.5)

    ax1.set_xlim(left=0.5, right=len(excess_returns_means_dist)+0.5)
    ax1.set_xticks(excess_returns_means_dist.index)
    ax1.set_xticklabels([int(x) for x in ax1.get_xticks()], fontproperties=font, fontsize=14)
    ax1.set_yticklabels([str(x*100)+'0%' for x in ax1.get_yticks()], fontproperties=font, fontsize=14)
    ax1.set_title(title, fontproperties=font, fontsize=16)
    ax1.grid(True)
```

```python
# std_turnover
fac_full = getTurnoverRateAll(universe=universe, begin=begin_date, end=end_date)
fac_full['tradeDate'] = map(Date.toDateTime, map(DateTime.parseISO, fac_full['tradeDate']))
fac_full.set_index('tradeDate',inplace=True)
fac_full[fac_full < 0.005] = np.nan

fac1_full = pd.rolling_mean(fac_full, window=20)

std_turnover = pd.DataFrame(index=fac_full.index,columns=fac_full.columns)
for tradeDate in fac_full.index:
    std_turnover.ix[tradeDate] = fac_full.ix[tradeDate] - fac1_full.ix[tradeDate]
```

```python
# std_ret               
fac2_full = getForwardReturnsAll(universe=universe, begin=begin_date, end=end_date)
fac2_full['tradeDate'] = map(Date.toDateTime, map(DateTime.parseISO, fac2_full['tradeDate']))
fac2_full.set_index('tradeDate',inplace=True)
fac2_full[fac2_full > 0.095] = np.nan
fac2_full[fac2_full < -0.095] = np.nan

fac3_full = pd.rolling_mean(fac2_full, window=20)

std_ret = pd.DataFrame(index=fac2_full.index,columns=fac2_full.columns)
for tradeDate in fac2_full.index:
    std_ret.ix[tradeDate] = fac2_full.ix[tradeDate] - fac3_full.ix[tradeDate]
```

# 2. 市场分位观察

（T 日标准化换手率 与 T+1 日标准化涨跌幅的绝对值的关系）

---
- 假设市场没有发生信息泄漏，各组数据应持平，意味着 **换手率不能对涨跌幅做出任何预测**
- 从下图可以看出，**市场确实存在信息泄露 的现象** 

```python
excess_returns_plot(fac=std_turnover, periods=np.abs(std_ret), title=u"T 日标准化换手率与 T+1 日标准化涨跌幅绝对值的关系")
```

（T 日标准化换手率 与 T+1 日标准化涨跌幅的关系）
