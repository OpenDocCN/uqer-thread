```python
import numpy as np
import pandas as pd

from sklearn.decomposition import KernelPCA

from CAL.PyCAL import *
cal = Calendar('China.SSE')
```

```python
def get_historical_closes(ticker, start_date, end_date):
    # get the data for the tickers.  This will be a panel
    #p = DataAPI.MktFunddAdjGet(secID=ticker,beginDate=start_date,endDate=end_date,field='secID,tradeDate,closePrice',pandas="1") 
    p = DataAPI.MktEqudAdjGet(secID=ticker,beginDate=start_date,endDate=end_date,field=u"secID,tradeDate,closePrice",pandas="1") 
    p['tradeDate'] = pd.to_datetime(p['tradeDate']) #因为DataAPI所有日期都是以字符串形式存储的，因此此处需要转换成日期时间型,这里很重要！
    # rename the columns
    p.rename(columns={'tradeDate': 'Date', 
                      'closePrice': 'Close'}, inplace=True)
    # pivot each ticker to a column
    pivoted = p.pivot(index='Date', columns='secID')
    # and drop the one level on the columns
    pivoted.columns = pivoted.columns.droplevel(0)
    return pivoted
```

```python
start = '2015-06-02'                       # 回测起始时间
end = '2016-06-02'                         # 回测结束时间

#获取回测期间全部数据
lookback = 250
delta = '-'+str(lookback)+'B'
data_start=cal.advanceDate(start,delta,BizDayConvention.Preceding).strftime('%Y%m%d')

#to retrieve stock data with pandas
universe =  set_universe("HS300")
data = get_historical_closes(universe, data_start, end)
data.info()
```

```python
#the index data
hs300 = DataAPI.MktIdxdGet(tradeDate=u"",indexID=u"000300.ZICN",ticker=u"",beginDate=data_start,endDate=end,field=u"tradeDate,closeIndex",pandas="1")
hs300 = hs300.rename(columns={'tradeDate':'Date','closeIndex':'HS300'})
hs300['Date'] =  pd.to_datetime(hs300['Date']) #因为DataAPI所有日期都是以字符串形式存储的，因此此处需要转换成日期时间型,这里很重要！
hs300 = hs300.set_index('Date')#转换后index是DatetimeIndex 
hs300.info()
```

```python
data = pd.concat([data, hs300], axis=1)
data.head()
```

```python
#In real-world applications, you would use return data

returns = data.pct_change()*100.0  #the returns
returns.fillna(0, inplace = True)  #  将NaN置换为0
# If any of the values of percentage returns equal zero, set them to a small number (stops issues with QDA model in scikit-learn)
for i,x in enumerate(returns["HS300"]):
    if (abs(x) < 0.0001):
        returns["HS300"][i] = 0.0001
returns.head()
```

```python
hs300index = pd.DataFrame(returns.pop('HS300'))
hs300index.info()
```

```python
returns.info()
```

```python
#Applying PCA

scale_function = lambda x: (x - x.mean()) / x.std() #Usually, PCA works with normalized data sets.

pca = KernelPCA().fit(returns.apply(scale_function))
len(pca.lambdas_)   #The importance or explanatory power of each component is given by its Eigenvalue.These are found in an attribute of the KernelPCA object.
```

```python
pca.lambdas_[:10].round() #let us only have a look at the first 10 components. The tenth component already has almost negligible influence
```

```python
#We are mainly interested in the relative importance of each component, so we will normalize these values.
get_we = lambda x: x / x.sum()
get_we(pca.lambdas_)[:10] 
```

```python
get_we(pca.lambdas_)[:10].sum()
```

```python
#The first component already explains about 45% of the variability in the 300 time series. The first five components explain about 64% of the variability.
```

```python
#Constructing a PCA Index
#Next, we use PCA to construct a PCA (or factor) index over time and compare it with the original index. First, we have a PCA index with a single component only:
pca = KernelPCA(n_components=1).fit(returns.apply(scale_function))
hs300index['PCA_1'] = pca.transform(-returns)*100
hs300index.head()
```

```python
import matplotlib.pyplot as plt
hs300index.apply(scale_function).plot(figsize=(8, 4))
```

```python
data = data.fillna(method='bfill')  #  在列方向上以后一个值作为值赋给NaN
data.head()
```

```python
hs300_index = pd.DataFrame(data.pop('HS300'))
hs300_index.info()
```

```python
scale_function = lambda x: (x - x.mean()) / x.std()
pca = KernelPCA().fit(data.apply(scale_function))
len(pca.lambdas_)
```

```python
pca.lambdas_[:10].round()
```

```python
get_we = lambda x: x / x.sum()
get_we(pca.lambdas_)[:10]
```

```python
get_we(pca.lambdas_)[:5].sum()
```

```python
#The first component already explains about 71% of the variability in the 300 time series. The first five components explain about 93% of the variability
pca = KernelPCA(n_components=1).fit(data.apply(scale_function))
hs300_index['PCA_1'] = pca.transform(-data)
hs300_index.apply(scale_function).plot(figsize=(8, 4))
```

```python
#Let us see if we can improve the results by adding more components. To this end, we need to calculate a weighted average from the single resulting components:
pca = KernelPCA(n_components=5).fit(data.apply(scale_function))
pca_components = pca.transform(-data)
weights = get_we(pca.lambdas_)
hs300_index['PCA_5'] = np.dot(pca_components, weights)
hs300_index.apply(scale_function).plot(figsize=(8, 4))
```

```python
#The results  are  not that much better than before,so...
```

```python
#To this end, let us divide the total time frame into two subintervals. We can then implement an early and a late regression:
cut_date = '2015/7/1'
early_pca = hs300_index[hs300_index.index < cut_date]['PCA_1']
early_reg = np.polyval(np.polyfit(early_pca,
            hs300_index['HS300'][hs300_index.index < cut_date], 1),
            early_pca)
```

```python
late_pca = hs300_index[hs300_index.index >= cut_date]['PCA_1']
late_reg = np.polyval(np.polyfit(late_pca,
            hs300_index['HS300'][hs300_index.index >= cut_date], 1),
            late_pca)                               
```

```python
import matplotlib as mpl
list_of_datetimes = list(data.index)
mpl_dates = mpl.dates.date2num(list_of_datetimes)
```

```python
plt.figure(figsize=(8, 4))
plt.scatter(hs300_index['PCA_1'], hs300_index['HS300'], c=mpl_dates)
plt.plot(early_pca, early_reg, 'r', lw=3)
plt.plot(late_pca, late_reg, 'r', lw=3)
plt.grid(True)
plt.xlabel('PCA_1')
plt.ylabel('HS300')
plt.colorbar(ticks=mpl.dates.DayLocator(interval=250),
           format=mpl.dates.DateFormatter('%d %b %y'))
```
