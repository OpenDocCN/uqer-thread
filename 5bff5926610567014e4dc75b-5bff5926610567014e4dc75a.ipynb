```python
df1=DataAPI.MktStockFactorsOneDayGet(tradeDate=u"20180111",secID=u"",ticker=u"",field=u"ticker,secID,secShortName,NetProfitGrowRate,PE,ROE5",pandas="1")
df1['NetProfitGrowRate'] = df1['NetProfitGrowRate']*100
df1.head()
```

```python
df_industry = DataAPI.EquIndustryGet(secID=u"",ticker=u"",industryVersionCD=u"010303",industry=u"",
                                     industryID=u"",industryID1=u"",industryID2=u"",industryID3=u"",intoDate=u"20180111",field=u"",pandas="1")
df_industry = df_industry[['ticker','secID','secShortName','industryID1','industryName1']]
df_industry.head()
```

```python
import pandas as pd #导入pandas函数集
df = pd.merge(df1,df_industry,how='left',on='ticker') #不知道怎么能把两个表中列名一样的不重合
del df['secID_y']  #删掉其中一个重合的 
df=df.rename(columns={'secID_x':'secID'} )   #将另一个改为原名
df.index=df.index + 1  #强迫症必备代码
df.head()
```

```python
df_industry1 = df.groupby('industryName1').agg({'PE':'mean'})
df_industry1
df_industry1=df_industry1.rename(columns ={'PE':'Average PE'} )
df_industry1.head()
```
