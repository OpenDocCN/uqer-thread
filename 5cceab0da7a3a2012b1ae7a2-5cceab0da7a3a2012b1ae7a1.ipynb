# 万圣节效应与MACD结合应用

* MACD（Moving Average Convergence and Divergence)是Geral Appel于1970年提出的，利用收盘价的短期（常用为12日）指数移动平均线与长期（常用为26日）指数移动平均线之间的聚合与分离状况，对买进、卖出时机作出研判的技术指标。由快的指数移动平均线（EMA12）减去慢的指数移动平均线（EMA26）得到快线DIF，由DIF的9日加权移动均线得到慢线DEA，再用2×（快线DIF-慢线DEA）得到MACD柱。
* MACD为正，说明DIF在DEA之上；MACD为负，说明DIF在DEA之下。所以当MACD由负转正时，即快线DIF上穿DEA，是买入的信号，反之当MACD由正转负时则是卖出的信号。

从上面两个策略对比来看，年化收益率差别不大，但在收益波动率和最大回撤方面有较大的改善，也就是说在享受差不多的收益时，风险降低了。
