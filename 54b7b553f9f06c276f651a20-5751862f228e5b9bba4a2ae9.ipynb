# 业绩基准类-分享版

此代码实现的功能主要有三个：

>* 利用DataAPI获得某个指数收益，作为基准

>* 将用户上传的收益率数据与所选的基准收益比较

>* 计算差值（用户收益-基准收益）的均值、波动率、夏普比率

```python
#将收益率数据读进来
rtn = pd.read_csv('return.csv',index_col=0,parse_dates=True)
Return = DataAPI.MktIdxdGet (indexID='000001.ZICN',field='CHGPct')
bench_rtn = pd.DataFrame(list(Return['CHGPct']),index=Return['tradeDate'],columns=['bench_return'])
merge_inf = rtn.join(bench_rtn, how='inner')

merge_inf['cm_rtn_bench'] = merge_inf['bench_return'].cumsum()    #计算基准累计收益
merge_inf['cm_rtn'] = merge_inf['return'].cumsum()    #计算策略的累计收益
#画图
merge_inf['cm_rtn'].plot()
merge_inf['cm_rtn_bench'].plot()
pylab.legend(loc=9)
```

```python
diff = merge_inf['return']-merge_inf['bench_return']    #差值

#print pylab.title('相对收益')
rtn_diff_avg = np.mean(diff)    #计算差值的均值
print '相对收益平均值：',rtn_diff_avg

volitality = np.sqrt(250)*np.std(diff)    #计算波动率，即年化标准差
print '相对收益波动率：',volitality

sharpe = rtn_diff_avg*240/volitality    #计算夏普比率
print '相对夏普比率：',sharpe
```
