# 单因子研究分析

本文试图以PE、PB、LFLO（对数流通市值）为例，探讨一下如何进行单因子分析的。测试的时间跨度是2009年到2016年，每月月初进行换股分组，分组的依据是当月初因子值从小到大排列，分为10组。
最后将这8年来的收益率与bench一起按照组别，每月相乘，得到总收益率，再进行年化，最后计算超额收益率。

一、引入必要的库

```python
import pandas as pd
from pandas import Series, DataFrame
import numpy as np
import statsmodels.api as sm
import scipy.stats as scs
import matplotlib.pyplot as plt
import time
from CAL.PyCAL import *
import datetime as dt
pd.set_option('display.float_format', lambda x: '%.3f' % x)
cal = Calendar('China.SSE')
from datetime import datetime

universe=set_universe('A')
n_quantile = 10#默认分为10组
```

二、选择研究的因子

```python
factors = [ 'PE',  'PB','LFLO' ]
```

三、定义函数。

```python
#月初取出因子数值
def get_factors(startdate,factors):
    fdate = Date.parseISO(startdate)
    if not cal.isBizDay(fdate):
        fdate=cal.advanceDate(fdate, '1B', BizDayConvention.Following)  
    fdf=DataAPI.MktStockFactorsOneDayGet(tradeDate=fdate,secID=universe,field=[u'secID']+factors,pandas="1").set_index('secID')
    df=pd.DataFrame()
    for factor in factors:
        ep = fdf[factor].dropna().to_dict()
        ep = standardize(neutralize(winsorize(ep),fdate.strftime('%Y%m%d')))#中性化、去极化、标准化
        ep=pd.DataFrame(ep.items(), columns=['secID', factor])
        ep=ep.dropna()
        if df.empty:
            df=ep
        else:
            df=pd.merge(df,ep)
    df=df.set_index('secID')
    return df

#获取每月月初工作日的函数
def BizMonthStart(startdate,enddate):
    data=pd.date_range(startdate,enddate,freq='BMS')
    data=[Date.fromDateTime(x) for x in data]
    date_list=[]
    for day in data:
        if (not cal.isBizDay(day)):
            day=cal.advanceDate(day, '1B', BizDayConvention.Following)
        date_list.append(day)
    date_list=[x.toISO() for x in date_list]    
    return date_list

#获取当月月末
def monthend(day):
    df=DataAPI.TradeCalGet(exchangeCD=u"XSHG",beginDate=day,field=['calendarDate','isMonthEnd'],pandas="1")
    df=df[df['isMonthEnd']==1]
    return df.iloc[0,0]

#获取每月月末工作日的函数
def BizMonthEnd(startdate,enddate):
    data=DataAPI.TradeCalGet(exchangeCD=u"XSHG",beginDate=startdate,endDate=enddate,field=['calendarDate','isMonthEnd'],pandas="1")
    data = data[data['isMonthEnd'] == 1]
    date_list = data['calendarDate'].values.tolist()
    return date_list

#计算月度收益
def caculate_port_monthly_return(port,startdate):
    enddate=monthend(startdate)
    chgpct=DataAPI.MktEqumGet(secID=port,monthEndDate=enddate,field=u"secID,chgPct",pandas="1")
    chgpct['chgPct']=chgpct['chgPct']+1
    negMarketValue=DataAPI.MktEqudGet(tradeDate=enddate,secID=port,field=u"secID,negMarketValue",pandas="1")
    df=pd.merge(chgpct,negMarketValue)
    Return=np.average(df['chgPct'],weights=df['negMarketValue'])-1
    return Return

#计算benchmark月收益
def caculate_benchmark_monthly_return(startdate,indexID):
    enddate=monthend(startdate)
    chgpct=DataAPI.MktIdxmGet(beginDate=startdate,endDate=enddate,indexID=indexID,field=u"endDate,chgPct",pandas="1")
    return chgpct.iloc[0,1]

#计算单因子的值
def calculate(startdate,enddate,factor,indexID='000001.ZICN'):
    MonthStart=BizMonthStart(startdate,enddate)
    df=pd.DataFrame()
    n=(pd.Period(enddate,'M')-pd.Period(startdate,'M')+1)
    for day in MonthStart:
        MonthEnd=monthend(day)
        #当月因子值
        fdf=get_factors(day,factors)
        fdf=fdf[factor]
        score = fdf.order()
       #分组
        cols = [i+1 for i in range(n_quantile)]
        pct_quantiles = 1.0/n_quantile
        
        for i in range(n_quantile):
            down = score.quantile(pct_quantiles*i)#因子临界值的下限
            up = score.quantile(pct_quantiles*(i+1))#因子临界值的上限
            port = score[(score<=up) & (score>=down)].index#处于上下界的股票代码列表
            
            portReturn=caculate_port_monthly_return(port,day)+1#当月收益率
            df.ix['port'+str(i+1),day]=portReturn
        benchReturn=caculate_benchmark_monthly_return(day,indexID)+1#当月基准收益率
        df.ix['bench',day]=benchReturn
        
    df['totalReturn']=df.T.cumprod().iloc[-1,:]
    df['annualReturn'] = (df['totalReturn'])**(12.0/n)
    df['excessReturn'] = df['annualReturn']- df['annualReturn'][-1]
    return df
    

#计算所有因子的值
def calculateFactors(startdate,enddate,indexID='000001.ZICN'):
    secs_time = 0
    start_time = time.time()
    result = {}

    for factor in factors:
        print str(factor)
        df=calculate(startdate,enddate,factor,indexID)
        result[factor]=df
        print str(factor)+' is ok'
        finish_time = time.time()#结束时间
        print '  ' + str(np.round((finish_time-start_time) - secs_time, 0)) + ' seconds elapsed.'#耗时
        secs_time = (finish_time-start_time)#调整已用时间
    monthly_return = pd.Panel(result)
    return monthly_return#是个panel
```

四、执行操作。因为这是全A股操作，所以很容易出现内存不足的问题，内存不足1G的朋友慎用。

```python
df=calculateFactors('2009-01-01','2016-12-31',indexID='000001.ZICN')
```
