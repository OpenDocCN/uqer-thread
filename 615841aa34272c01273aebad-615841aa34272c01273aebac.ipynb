# 场外基金因子合成及买卖实证

# 原理
## 首先筛选基金
条件有：
- 基金池为中证主动基金指数成分基金，该指数的成分基金为排除指数基金、指数增强基金、ETF、LOF后的股票或混合基金。
- 成立1年3个月以上（3个月建仓期，1年反应业绩）
- 最近一期规模大于5亿。
- 最近4个季度股票占比大于60%。
- 最近4个季度重仓股变动比例小于70%。
## 每季度从筛选出的基金中按各种因子计算得分
分值因子：
- 基金规模（按净值计算），这项越小越好。
- 一年夏普比率
- 机构持有比率
- FF三因子+TM模型回归计算的alpha
计算方法：取市场收益率RM、大市值与小市值收益率差SMB、高估值与低估值收益率差HML，按以下公式计算:
$R=\alpha+\beta_1R_m+\beta_2R_m^2+\beta_{HML}R_{HML}+\beta_{SMB}R_{SMB}$
以上因子按MAD去极值后计算标准分，等权相加为基金得分。

```python
import numpy as np
import pandas as pd
import pandas as pd
import numpy as np
from scipy.optimize import minimize 
import statsmodels.api as sm
from lib.TradeDate import *
from datetime import timedelta
from datetime import datetime

#=========常用函数========================#
#OLS回归
def my_ols(dataX,dataY):
    X=sm.add_constant(dataX)
    model = sm.OLS(dataY,X).fit()
    return model.params
#计算标准分
#ascending=False,值越大分越高。True值越小分越高
def z_score(data,ascending=False):
    data=data.dropna()
    data=filter_extreme_MAD(data,5)
    mean=data.mean()
    std=data.std()
    if not ascending:
        data=(data-mean)/std
    else:
        data=(mean-data)/std
    return data
#中位数去极值
def filter_extreme_MAD(data,n): #MAD: 中位数去极值  
    median = data.quantile(0.5)  
    new_median = ((data - median).abs()).quantile(0.50)
    max_range = median + n*new_median  
    min_range = median - n*new_median  
    #print(median,new_median,max_range)
    data[data>max_range]=max_range
    data[data<min_range]=min_range
    return data 
#N月前日期
def date_start(end_date,monthCount):
    days=monthCount*30
    deltaDay=timedelta(days=days)
    start_date=(datetime.strptime(end_date,'%Y%m%d')-deltaDay).strftime("%Y-%m-%d")
    return start_date
#========获取基金因子=====================#
# 获取单因子值
def get_factor(ticker,window,factor,tradeDate):
    data=DataAPI.FundPerfIndicGet(dataDate=tradeDate,ticker=ticker,window=window,
                                  field=u"ticker,"+factor,pandas="1")
    data.set_index('ticker',inplace=True)
    return data

# 获取机构持有比例
def get_holding_ratio(ticker,tradeDate):
    start_date=date_start(tradeDate,6)
    data=DataAPI.FundHolderInfoGet(ticker=ticker,beginDate=start_date,
                                    endDate=tradeDate,
                                    field=u"ticker,instHoldRatio",
                                    pandas="1")
    data.set_index('ticker',inplace=True)
    return data

# 最近规模
def fund_asset(ticker,end_date):
    start_date=date_start(end_date,3)
    data=DataAPI.FundAssetsGet(ticker=ticker,beginDate=start_date,endDate=end_date,
                               field=u"ticker,netAsset",pandas="1")
    data.set_index('ticker',inplace=True)
    return data

# 最近4季度股票占比
def equ_ratio(ticker,end_date,equity_ratio=60):
    start_date=date_start(end_date,12)
    data=DataAPI.FundAssetsGet(ticker=ticker,beginDate=start_date,endDate=end_date,
                               field=u"ticker,equityRatioInTa",pandas="1")
    data=data.groupby(['ticker']).min()
    data=data[data['equityRatioInTa']>equity_ratio]   
    ticker=list(data.index)
    return ticker

# 最近4季度十大重仓股变动最大值，输入change_ratio为整型
def equ_change(ticker,end_date,change_ratio=3):
    start_date=date_start(end_date,12)
    data=DataAPI.FundHoldingsGet(ticker=ticker,beginDate=start_date,endDate=end_date,secType="E"
                                 ,field=u"ticker,reportDate,holdingSecID,marketValue",pandas="1")
    data.sort_values(by='marketValue',inplace=True,ascending=False)
    data=data.groupby(['ticker','reportDate']).head(10)
    # 对比两个季度持股差别
    change={}
    for sec in ticker:
        tmpData=data[data['ticker']==sec]
        reportDate=list(tmpData['reportDate'].drop_duplicates())
        result=[]
        for i in range(1,len(reportDate)):
            old_stock=list(tmpData[tmpData['reportDate']==reportDate[i-1]]['holdingSecID'])
            new_stock=list(tmpData[tmpData['reportDate']==reportDate[i]]['holdingSecID'])
            result.append(len(set(old_stock)&set(new_stock)))
        change[sec]=np.mean(result)
    change=pd.Series(change)
    ticker=list(change[change>change_ratio].index)
    return ticker

#========获取基金FF因子=====================#
#计算基金收益率序列
def get_fund_series(fund,start_date,end_date):
    data=DataAPI.FundNavGet(ticker=fund,beginDate=start_date,endDate=end_date,field=u"ticker,endDate,adjNavChgPct",pandas="1")
    data=data.groupby(['ticker','endDate']).mean().unstack().T
    data=data.loc['adjNavChgPct']
    data=data/100
    return data
# 获取基准收益率序列
def get_base_series(base,start_date,end_date):
    #计算市场收益
    dataBase=DataAPI.MktIdxdGet(ticker=base,field='tradeDate,CHGPct',beginDate=start_date,endDate=end_date,pandas='1')
    dataBase.set_index('tradeDate',inplace=True)
    dataBase['CHGPct']=dataBase['CHGPct']
    return dataBase['CHGPct']
#计算RM、SMB、HML序列
def get_ref_series(base,start_date,end_date):
    #计算市场收益率
    rm=get_base_series(base,start_date,end_date)
    #使用中证全指成分股
    secID=list(DataAPI.IdxConsGet(secID=base,intoDate=start_date,field=u"consID",pandas="1")['consID'])
    #获得起始日期所有股票流通市值和估值数据
    data=DataAPI.MktEqudGet(secID=secID,tradeDate=start_date,field=u"secID,negMarketValue,PB",pandas="1")
    data.set_index('secID',inplace=True)
    data.dropna(inplace=True)
    data=data[data['PB']>0]
    #将所有股票按市值分为S、B两组，每组按估值分为L、M、H三组。
    #计算SL\SM\SH的每组收益率均值的均值-BL\BM\BH的均值为市值风格收益率
    #计算SH\BH的均值-SL\BL的均值为估值风格收益率
    S=list(data.sort_values(by='negMarketValue',ascending=True).iloc[:int(len(data)/3)].index)
    SL=list(data[data.index.isin(S)].sort_values(by='PB',ascending=True).iloc[:int(len(S)/3)].index)
    SM=list(data[data.index.isin(S)].sort_values(by='PB',ascending=True).iloc[int(len(S)/3):int(len(S)/3*2)].index)
    SH=list(data[data.index.isin(S)].sort_values(by='PB',ascending=True).iloc[-int(len(S)/3):].index)
    B=list(data.sort_values(by='negMarketValue',ascending=True).iloc[-int(len(data)/3):].index)
    BL=list(data[data.index.isin(B)].sort_values(by='PB',ascending=True).iloc[:int(len(S)/3)].index)
    BM=list(data[data.index.isin(B)].sort_values(by='PB',ascending=True).iloc[int(len(S)/3):int(len(S)/3*2)].index)
    BH=list(data[data.index.isin(B)].sort_values(by='PB',ascending=True).iloc[-int(len(S)/3):].index)
    #计算样本时间内各因子收益率
    chgPct=DataAPI.EquRetudGet(secID=list(data.index),beginDate=start_date,endDate=end_date,
                               isChgPctl=u"0",field=u"secID,tradeDate,dailyReturnReinv",pandas="1")
    chgPct=chgPct.groupby(['secID','tradeDate']).mean().unstack()
    pSL=chgPct[chgPct.index.isin(SL)].mean(axis=0).loc['dailyReturnReinv']
    pSM=chgPct[chgPct.index.isin(SM)].mean(axis=0).loc['dailyReturnReinv']
    pSH=chgPct[chgPct.index.isin(SH)].mean(axis=0).loc['dailyReturnReinv']
    pBL=chgPct[chgPct.index.isin(BL)].mean(axis=0).loc['dailyReturnReinv']
    pBM=chgPct[chgPct.index.isin(BM)].mean(axis=0).loc['dailyReturnReinv']
    pBH=chgPct[chgPct.index.isin(BH)].mean(axis=0).loc['dailyReturnReinv']
    r_SMB=(pSL+pSM+pSH)/3-(pBL+pBM+pBH)/3
    r_HML=(pSH+pBH)/2-(pSL+pBL)/2
    return rm,r_SMB,r_HML
# 多元回归计算风格(FF+TM)
def cal_style_TM(r_fund,rm,r_SMB,r_HML):
    rm2=np.power(rm,2)
    result={}
    #整合市场、SMB、HML序列
    df_style=pd.concat([rm,rm2,r_SMB,r_HML],axis=1)
    df_style.columns=['rm','rm2','SMB','HML']
    #回归分析
    res=my_ols(df_style,r_fund)
    result['alpha']=round(res['const']*100,2)
    result['RM2']=round(res['rm2'],2)
    result['RM']=round(res['rm'],2)
    result['SMB']=round(res['SMB'],2)
    result['HML']=round(res['HML'],2)
    return result
# 计算FF因子主函数
def fund_FF_factor(fund,end_date,period,base='000985'):
    result={}
    start_date=get_tradedate_before(end_date,period)[0]
    r_fund=get_fund_series(fund,start_date,end_date)
    rm,r_SMB,r_HML=get_ref_series(base,start_date,end_date)
    #去除基金净值中的异常日期
    r_fund=r_fund[r_fund.index.isin(list(rm.index))]
    for f in fund:
        r=cal_style_TM(r_fund[f],rm,r_SMB,r_HML)
        result[f]=r
    return pd.DataFrame(result).T

#========以下是核心函数===============

#初步筛选基金，注意传入日期格式"%Y%m%d"
def filter_fund(tradeDate):
    #获取所有混合基金,数据量大
    #funds1=DataAPI.FundGet(category=u"H",operationMode=u"O",endDate=tradeDate,field=u"",pandas="1")
    funds1=DataAPI.IdxConsGet(ticker=u"930890",intoDate=tradeDate,field=u"consID,consTickerSymbol",pandas="1")
    ticker=list(funds1.consTickerSymbol)
    #去掉名字重复的
    funds=DataAPI.FundGet(ticker=ticker,field=u"ticker,secShortName",pandas="1")
    funds['name6']=funds['secShortName'].str[:18]
    funds=funds.drop_duplicates('name6')
    ticker=list(funds.ticker)
    #去掉指数基金和指数增强基金
    #funds1=funds1[~ ((funds1['indexFund']=='I') | (funds1['indexFund']=='EI'))]
    #funds1=funds1[funds1['isQdii']!=1]
    #funds1=funds1[(funds1['secID'].str.contains('OFCN')) | (funds1['secID'].str.contains('XSHG')) | (funds1['secID'].str.contains('XSHE'))]
    funds1=DataAPI.FundGet(ticker=ticker,field=u"",pandas="1")
    #print(datetime.now(),'筛选成立时间')
    #成立时间大于1年零3个月
    monthCount=15
    start_date=date_start(tradeDate,monthCount)
    funds1=funds1[funds1.establishDate<start_date]
    ticker=list(funds1['ticker'])
    #print(datetime.now(),'筛选规模')
    #最近规模大于5亿
    asset=fund_asset(ticker,tradeDate)
    ticker=list(asset[asset['netAsset']>5e+08].index)
    #print(datetime.now(),'筛选股票比例')
    #最近四季度股票占比大于60%
    ticker=equ_ratio(ticker,tradeDate)
    #print(datetime.now(),'筛选重仓股变动')
    #最近4个季度重仓股变动比例<=70%
    ticker=equ_change(ticker,tradeDate)
    return ticker

#每季度调用，选择最优基金
def rebalance(tradeDate):
    ticker=filter_fund(tradeDate)
    period=120
    index_market='000985'#中证全指
    FF=fund_FF_factor(ticker,end_date,period)
    instHold=get_holding_ratio(ticker,tradeDate)
    sharpe=get_factor(ticker,5,'sharpeRatio',tradeDate)
    net_asset=fund_asset(ticker,tradeDate)
    sharpeScore=z_score(sharpe['sharpeRatio'])
    instHoldScore=z_score(instHold['instHoldRatio'])
    FFScore=z_score(pd.Series(FF['alpha']))
    assetScore=z_score(net_asset['netAsset'],ascending=True)
    score=sharpeScore+instHoldScore+FFScore+assetScore
    to_hold=list(score.sort_values(ascending=False).index)[:5]
    return to_hold

#=======主函数=========
from lib.functions import *
start_date='20170401'
end_date='20210901'
dates=get_month_last_tradeDate(start_date,end_date)
is_first=True#第一个日期不计算收益
holding=''
fund_result={}
for d in dates:
    d=d.replace('-','')
    if d[5] in ['4','7','10','1']:
        holding=rebalance(d)
        fund_result[d]=holding
```
