**Alpha#30**
$$(((1.0 - rank(((sign((close - delay(close, 1))) + sign((delay(close, 1) - delay(close, 2)))) +
sign((delay(close, 2) - delay(close, 3)))))) * sum(volume, 5)) / sum(volume, 20))$$

先了解下几个运算符：<br>
**Rank(x)** ： 排列所有工具的 x 值，然后返回的数值是介于0.0和1.0之间均匀分布的浮点数。<br>
例如假设你有一个向量：<br>
A = [29, 12, 5, 7, 10]<br>
现在，将会为向量里的每个元素**按升序分级**，最小一级为 0：<br>
A = [29, 12, 5, 7, 10]<br>
A' = [4, 3, 0, 1, 2] <br>
每个“等级”再除以（元素个数 - 1）。于是得到：<br>
Rank (A) = [1.00, 0.75, 0.00, 0.25, 0.50]<br>
所以您等到了一个从 0 到 1 间距相等的向量。<br>

**Sign(x)** : 	如果 x > 0，就返回 1，如果 x < 0，则返回 -1，如果 x == 0，返回 0

**Delay(x, n)** :  	n 天前 x 的值。请注意，n 必须小于256

首先观察公式，涉及的数据为收盘价与成交量。这里并为计算它们的相关系数，而是**分别从价和量的角度**去构造的alpha。（当然它们不是割裂的，最终还是综合到价量关系上）。

然后我们发现式子这一部分：$$ sign((close - delay(close, 1))) + sign((delay(close, 1) - delay(close, 2)))) +
sign((delay(close, 2) - delay(close, 3))$$
其实是分别计算前三天的收盘价相对他们之前一天的差。然后，对于大于零的，也就是上涨的我们标记为1；对于下跌的我们标记为-1；持平我们标记为0。 将他们求和后再进行rank操作。我们将这部分记为x1。

x1表示了上涨天数和下跌天数的力量。连续3天上涨 > 上涨两天持平1天 > 上涨2天下跌1天。。。 而式子的左边为 1-x1， 因此我们的偏好是相反的。 也就是，前3天跌的天数越多，我们分配更多权重。这是典型的**均值回归**思想。在其他条件未知的情况下，前面下跌天数越多，之后价格向上回归的可能越高。（当然，持有时间太短也许等不到反弹，因此调仓周期长一点可能性越高。）

式子右半部分则是过去5天成交量的和与过去20天成交量的比值。反映了近日成交量放大或缩小，显然，我们希望近日的成交量是放大的。结合式子坐半部分，我们要找的是前面3天跌率最高，且成交量放大的股票。

另外，这里个人认为与量增价跌有区别。量增价跌是成交量与价格变化方向相反，而这里强调的是下跌态势中成交量相对近20天是放大的。


可以看出该alpha是有效的，调仓周期太小的话，表现下降。
