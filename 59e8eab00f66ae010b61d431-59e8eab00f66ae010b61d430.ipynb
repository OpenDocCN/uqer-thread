# Python金融时间序列分析系列3之线性序列

今天 我们来讲一讲线性时间序列。我们先来介绍几个线性时间序列理论知识。

一般来说，分析一个统计学问题，都需要一些假设，而平稳性是时间序列分析的基础。这里的平稳性，我们可以理解是时间序列的行为并不随时间改变。一般来说，平稳性分为严平稳和弱平稳。严平稳是指如果对所有的时刻 t，任意正整数 n 和任意 n 个正整数 $$ \large (t_1,t_2......t_n), \\ \large (Z_{t_1},Z_{t_2},......Z_{t_n})$$的联合分布与 $$ \large (Z_{t_1-k},Z_{t_2-k},......Z_{t_n -k}) $$的联合分布相同，我们称时间序列  {$Z_t$} 是**严平稳**的。
但这是一个很强的条件，无论从实际上还是理论上，所以一般我们假设是弱平稳。弱平稳是指若时间序列 {$r_t$} 满足下面两个条件:$$ \large E(Z_t) = \mu, \mu 是常数\\ \large Cov(Z_t,Z_{t-l}) = \gamma_l, \gamma_l 只依赖于l $$
则时间序列 {$Z_t$} 为**弱平稳**的。

**相关系数**
两个变量的相关系数定义如下：
  $$ \large \rho_{xy} = \frac{Cov(X,Y)}{\sqrt{Var(X)Var(Y)}} =   \frac{E{[(X-\mu_x)(Y-\mu_y)]}}{\sqrt{E{(X-\mu_x)^{2}}E{(Y-\mu_y)^{2}}}} $$
其中 $$ \mu_x 和 \mu_y 是变量X和Y的均值，并且假定变量的方差存在。相关系数衡量的是X和Y线性相关的程度。可以证明  -1 \leqslant  \rho_{xy} \leqslant 1  并且 \rho_{xy} = \rho_{xy}$$
$$若 \rho_{xy} = -1或1,         X和Y完全线性相关 $$
$$若 -1 < \rho_{xy} < 1 ,       X和Y完全线性相关 $$
$$若 \rho_{xy} = -0      ,   X和Y完全线性无关 $$

**自相关函数（ACF）**
当我们考虑到$Z_t$ 与它的过去值 $Z_{t-l}$ 的线性相关性。 这时候我们把相关系数的概念推广到自相关系数。
$Z_t$ 与 $Z_{t-l}$ 的相关系数称为 $Z_t$ 的**间隔为 l 的自相关系数，通常记为 $ρ_l$ 。具体的：**
$$\large \rho_l = \frac{Cov(Z_t,Z_{t-l})} {\sqrt{Var(Z_t)Var(Z_{t-l})}} = \frac{Cov(Z_t,Z_{t-l})} {Var(Z_t)}  = \frac{\gamma_l}{\gamma_0}$$
** 这里用到了弱平稳序列的性质:** $$ \large Var(Z_t)=Var(Z_{t-l})$$ 由定义我们知道$$\large \gamma_0 = 1,\gamma_l=\gamma_{-l}和 -1 \leqslant  \gamma_l \leqslant 1$$


对一个**平稳时间序列**的样本 {$Z_t$}， 1 <= t <= T,则间隔为 l 的样本自相关系数的估计为:
$$ \large \hat \rho_l = \frac{\sum_{t=l+1}^{T}(Z_t- \bar Z)(Z_{t-l}-\bar Z)}{ \sum_{t=1}^{T}(Z_t- \bar Z)^{2}}, 0 \leqslant l \leqslant T-1 $$

则函数 $$ \large \hat \rho_1,\hat \rho_2 ,\hat \rho_3...$$ 称为 $r_t$ 的**样本自相关函数(ACF)**

另外一个弱平稳序列是序列不相关的当且仅当所有**所有的$ \large l>0$**,有 $\gamma_l = 0$,因此，我们经常需要检验多个自相关系数是否为0。

**混成检验**
为了检验多个自相关系数是否为0，我么这里需要用到混成检验。

原假设 $$ H_0: \rho_1 = ... = \rho_m=0 $$
备择假设 $$ H_a: \exists i \in \{1,...,m\}, \rho_i \ne 0 $$

混成检验统计量：$$\large Q(m) = T(T+2) \sum_{l=1}^{m} \frac{\hat{\rho_l}^{2}}{T-l}$$
Q(m)渐进服从自由度为m的 $$ \chi^2 $$分布
决策规则：
$$ \large Q(m) > \chi_\alpha^{2} ,拒绝H0$$
即,Q(m)的值大于自由度为m的卡方分布100(1-α)分位点时，我们拒绝H0。

**大部分软件会给出Q(m)的 p-value，则当 p-value 小于等于显著性水平 α时拒绝H0。**

接下来来看一个实例

```python
from scipy import  stats
import numpy as np
import pandas as pd
import statsmodels.api as sm  # 统计相关的库

df =DataAPI.MktEqudGet(ticker=u"000001",beginDate=u"20170701",endDate=u"20170801",field=u"TradeDate,closePrice",pandas="1")
data = df['closePrice'] # 上证指数
m = 10 # 我们检验10个自相关系数

acf,q,p = sm.tsa.acf(data,nlags=m,qstat=True)  ## 计算自相关系数 及p-value
result = np.c_[range(1,11), acf[1:], q, p]
result_pt=pd.DataFrame(result, columns=["ix" ,"AC", "Q", "P-value"])
result_pt = result_pt.set_index('ix')
result_pt
```

显著性水平α，是在原假设成立时，检验统计量值落在某个极端区域的概率值。因此，若取α= 0.05，如果计算出的p值小于α ，则可认为原假设是一个不可能发生的小概率事件。当然，如果真的发生了，则犯错误的可能性为5%。显然，显著性水平反映了拒绝某一原假设时所犯错误的可能性，或者说，α是指拒绝了事实上正确的原假设的概率。
一般最常用的0.05和0.01，医学类大多采用0.01。所以这里取显著性水平为0.05，可以看出，所有的p-value都小于0.05；因此则我们拒绝原假设$H_0$。

因此，我们认为该序列是**序列相关**的。
