# 一.前言

- ### 传统多因子模型忽视了个股之间的基本面情况差异和选股因子在不同风格股票池里的适用性，因此产生了能够捕捉不同股票之间差异性的动态情景模型(Dynamic Contextual Alpha Model)
- ### 我们知道，不同个股的基本面属性可能存在非常大的区别，所以传统的基本面研究以行业或者市值作为划分标准的依据。人们尝试寻找在行业内有显著效果的因子对行业内的股票进行独立打分，然后再汇总个股的得分或者排序，从而得到全市场的排序。这一方法提高了Alpha模型的广度(Breadth),但也存在几个问题。比如，行业的分类主要依据是主营业务，所以公司的基本面并不一致；随着公司的成长，基本面也在发生着变化；行业内的股票有限，容易造成过拟合。
- ### 动态情景Alpha模型摒弃了行业内打分的做法，转而根据股票基本面的属性，比如估值,成长和盈利能力等，把股票分成不同的层面，对每个层面内的股票采用单独的评价体系进行打分或者排序，最后得到每个股票的综合得分。这种方法能有效的捕捉市场的风格切换，提升因子模型的适应能力。
- ### 本帖按照研报<<动态情景多因子Alpha模型----因子选股系列研究之八>>完成前期情景因子分层的部分工作。

# 二.因子分类
- ### 既然要对因子进行分层，首先就要将因子进行分类，比如从风格维度上来说，因子可以分为规模，价值，成长，盈利，流动性水平等。
- ### 优矿的DataAPI里面提供了因子的数据和分类，为了方便今后使用因子，不妨按照优矿的分类将因子整合一下

```python
#加载需要使用的包
import numpy as np
import pandas as pd
import scipy.stats as st
```

- ### 先将DataAPI.MktStockFactorsOneDayGet的详情描述复制粘贴存为excel，再把这个excel上传到研究环境的data里

```python
#对DataAPI.MktStockFactorsOneDayGet里的因子进行分类
factorDescription = pd.read_excel('factorDescription.xlsx')    #把DataAPI.MktStockFactorsOneDayGet的描述详情存为excel
del factorDescription[u'类型']
for i in range(len(factorDescription)):     #提取对因子描述里的关键字，即’属于’后面的文字
    try:                                    
        tmp = factorDescription[u'描述'][i]
        factorDescription[u'描述'][i] = tmp[tmp.index(u'。属于')-len(tmp)+3:]
        if u'。' in factorDescription[u'描述'][i]:
            factorDescription[u'描述'][i] = factorDescription[u'描述'][i][:-1]
    except:
        factorDescription[u'描述'][i]='null'         #对于描述文字里没有归类的因子，舍弃不管
classes = factorDescription[u'描述'].drop_duplicates().tolist()   #得到因子类别
classify={}            #字典，存储因子类别及因子信息
for t in classes:
    if t == 'null':
        continue
    classify[t] = ','.join(factorDescription[factorDescription[u'描述'] == t][u'名称'].tolist())
result = pd.DataFrame(index=classify.keys(),columns=[u'因子名称'],data=0)
result[u'因子名称'][classify.keys()] = classify.values()    #把字典传化为dataframe
result
```

- ### 这样或许还不是很方便查找自己想要的因子，不妨写个函数，输入中文即可查找相关因子，满足像ctrl+F的功能

```python
def find_inf(zhongwen):
    factorDescription = pd.read_excel('factorDescription.xlsx')  #读取数据
    del factorDescription[u'类型']
    indexs=[]
    for i in range(len(factorDescription)):           
        if zhongwen in factorDescription[u'描述'][i]:   #寻找关键字所在的行
            indexs.append(i)
    return factorDescription.iloc[indexs,:]             #输出所有关键字所在的行
```

# 三.情景分层因子的构建
- ### 模型构建的基础是找到用什么因子对股票进行分层，从而把握市场的风格。分层的目标是把具有相似的基本面的股票聚集在一起，同时不同的分层又能够刻画股票基本面上不同的属性。分层的理性结果是不同的Alpha因子在不同的分层(如大市值和小市值，高估值和低估值)有着显著的绩效差别。
- ### 鉴于以上目标和平时研究因子性质的需要，我把整个过程写成了如下的类，处理的问题就是看一段时间内因子在所指定因子分层下的表现。
- ### 具体的讲，就是：
> ### 1.指定一个分层因子和其它被观察的因子，以及研究的时间范围，数据频率（如周度,月度等），股票范围
> ### 2.在每个规定的时间点上，首先把想要研究的全部股票按照分层因子的值等分成两块(high,low)，每一块的股票数量相同，如大市值股票和小市值股票
> ### 3.计算其它因子在不同股票池上的风险调整后的Rank IC ,和IC_IR
> ### 4.利用统计检验分析因子IC在不同分层下的显著性差异
- ### 这里的风险调整IC计算公式为:$IC_{adj}=corr(f_{pure},r_{residual})$,其中$f_{pure}$是上期因子值中性化后的值，$r_{residual}$是当期个股收益中性化后的结果，回归的方程表示为：
$$f_{pure}=f-b_1X-b_2log(mktcap)$$
$$r_{residual}=r-m_1X-m_2log(mktcap)$$
$$r为当期个股收益,f为上期因子原始值,X为行业虚变量，log(mktcap)为市值的对数$$
- ### 在优矿上，可以用函数neutralize来代替回归这一步
- ### IC_IR就是IC值得均值除以方差
- ### 这里的统计检验，我使用了三个，一个是针对均值的t检验，一个是针对方差的检验levene检验，和对分布的K-S检验

```python
class factor_perf_on_basical_factor(object):
    
    def __init__(self,basical_factor,factor,begindate,enddate,frequency='monthly',universe_name='A',dynamic=False):
        ###----传入参数----
        self.begindate = begindate  #开始时间
        self.enddate = enddate      #结束时间
        self.frequency = frequency  #数据频率，只要日度，周度，月度三个选择
        self.universe_name = universe_name  #股票收益溢价的股票池
        self.basical_factor = basical_factor #基本因子，用于分层
        self.factor = factor   #想要查看因子表现的因子名
        self.dynamic = dynamic   #universe是否使用动态
        ###----自定义参数----

        self.datelist = self.get_datelist()  #传入开始和结束日期，得到交易日列表list
        
    
    ###得到日期列表
    def get_datelist(self):   
        from CAL.PyCAL import *  #从dataapi调取交易日序列
        data=DataAPI.TradeCalGet(exchangeCD=u"XSHG",beginDate=self.begindate,endDate=self.enddate,field=['calendarDate','isWeekEnd','isMonthEnd'])
        if self.frequency == 'daily':   #按照规定的频率处理得到的交易日序列
            #self.M = 250
            date= map(lambda x: x[0:4]+x[5:7]+x[8:10], data['calendarDate'].values.tolist())
        if self.frequency == 'weekly':
            #self.M = 52
            data = data[data['isWeekEnd'] == 1]
            date= map(lambda x: x[0:4]+x[5:7]+x[8:10], data['calendarDate'].values.tolist())
        if self.frequency == 'monthly':
            #self.M = 12
            data = data[data['isMonthEnd'] == 1]
            date= map(lambda x: x[0:4]+x[5:7]+x[8:10], data['calendarDate'].values.tolist())
        return date
    
    #按basical_factor把股票分为数量相同的两组（大小），传入分组时的时间点
    def get_group(self,breakpoint):
        factor_filed = 'secID,'+self.basical_factor
        if self.dynamic:          #调用dataAPI得到股票代码及股票分层因子的值，如果是动态universe,则使用动态
            data = DataAPI.MktStockFactorsOneDayGet(tradeDate=breakpoint,secID=set_universe(self.universe_name,breakpoint),field=factor_filed).dropna()
        else:    
            data = DataAPI.MktStockFactorsOneDayGet(tradeDate=breakpoint,secID=set_universe(self.universe_name),field=factor_filed).dropna()
        data.sort(data.columns[1],ascending=True,inplace=True)     #按分层因子值从小到大排序
        secIDs = data['secID'].tolist()
        group_low = secIDs[:np.round(len(data))/2]        #分组,因子值小的哪一组股票为low,高的为high
        group_high = secIDs[np.round(len(data))/2:]
        return group_low,group_high
    
    #定义函数计算股票收益序列，传入时间，股票代码，得到收益序列
    def get_return(self,secIDs,date):
        index = self.datelist.index(date)
        data1 = DataAPI.MktEqudAdjGet(tradeDate=self.datelist[index-1],secID=secIDs,field=u"secID,closePrice").dropna() #前一期股票收盘价
        data1.set_index('secID',inplace=True)
        data2 = DataAPI.MktEqudAdjGet(tradeDate=date,secID=secIDs,field=u"secID,closePrice").dropna()  #当期股票收盘价
        data2.set_index('secID',inplace=True)
        data = pd.concat([data1,data2],axis=1).dropna()
        data.columns=['return','closeprice']
        data['return'] = data['closeprice']*1.0/data['return']-1    #得到收益序列
        del data['closeprice']
        returns = standardize(neutralize(winsorize(data['return'].dropna().to_dict()),date))    #对个股收益作标准化，去极值，中性化处理
        data['return'][returns.keys()] = returns.values()
        return data
    
    #得到因子值
    def get_factor(self,secIDs,date):
        date = self.datelist[self.datelist.index(date)-1]
        factor_filed = 'secID,'+self.factor
        data = DataAPI.MktStockFactorsOneDayGet(tradeDate=date,secID=secIDs,field=factor_filed).dropna()  #得到上期因子值
        data.set_index('secID',inplace=True)      #设置index为股票代码
        for i in data.columns:
            tmp = standardize(neutralize(winsorize(data[i].to_dict()),date))  #对因子值作标准化，去极值，中性化处理
            data[i][tmp.keys()] = tmp.values()
        return data
    
    #计算因子的rank_IC
    def get_rankIC(self):
        low={}
        high={}
        tmp = DataAPI.MktStockFactorsOneDayGet(tradeDate='20160930',ticker='000001',field=self.factor)
        for i in tmp.columns:
            low[i] = []                    #用于存储low那一组股票的rankIC
            high[i] = []                    #用于存储high那一组股票的rankIC
        for j in range(1,len(self.datelist)):   #对时间做循环，得到每个时间点的rankIC
            date = self.datelist[j]
            group_low,group_high = self.get_group(date)         #分组
            returns_low = self.get_return(group_low,date)
            returns_high = self.get_return(group_high,date)     #得到当期收益序列
            factor_low = self.get_factor(group_low,date)
            factor_high = self.get_factor(group_high,date)      #得到上期因子序列
            returns_low = pd.concat([returns_low,factor_low],axis=1).dropna()    #对应起来
            returns_high = pd.concat([returns_high,factor_high],axis=1).dropna()
            for t in tmp.columns:
                tmplow,_=st.spearmanr(returns_low['return'],returns_low[t])
                tmphigh,_=st.spearmanr(returns_high['return'],returns_high[t])
                low[t].append(tmplow)
                high[t].append(tmphigh)
        return low,high,tmp.columns
    
    #分析因子rank_IC
    def analysis(self):
        low,high,column_names = self.get_rankIC()
        result = pd.DataFrame(columns=column_names,index=np.arange(12),data=0)
        for i in column_names:
            mean_high=np.array(high[i]).mean()  #计算均值
            mean_low=np.array(low[i]).mean()
            std_high=np.array(high[i]).std()    #计算方差
            std_low=np.array(low[i]).std()
            t,p_t=st.ttest_ind(high[i], low[i], equal_var=False)   #t检验
            F,p_F=st.levene(high[i], low[i])                       #关于方差的F检验
            ks,p_ks=st.ks_2samp(high[i], low[i])                   #关于分布的K-S检验
            result[i] = [mean_high,mean_low,std_high,std_low,mean_high/std_high,mean_low/std_low,t,p_t,F,p_F,ks,p_ks]
        result = result.T
        np.arrays = [['mean','mean','std','std','IR','IR','Two sample t test','Two sample t test','levene test','levene test','K-S test','K-S test'],['high','low','high','low','high','low','t','p_value','F','p_value','KS','p_value']]
        result.columns = pd.MultiIndex.from_tuples(zip(*np.arrays))   #生成多维列名
        return pd.concat([result],keys=[self.basical_factor+' 分层后因子表现    时间范围：'+self.begindate+'--'+self.enddate+'    数据频率：'+self.frequency+'  股票范围:'+self.universe_name],axis=1)
```

- #### 尝试一下这个函数是否work

```python
test = factor_perf_on_basical_factor('LCAP','PE,ASSI','20120901','20140930')
result_test = test.analysis()
result_test
```

- ### 先解释一下上述结果，首先表格第一行给出了本次检验的基本信息,分层因子，时间范围，数据频率和股票范围
- ### 第二，三行是对下面数据的说明，比如mean表示均值，而high和low则分别表示分层因子值高的股票组合和分层因子值低的股票组合。
- ### t检验的原假设是两个样本的均值相同，所以p值越低,说明两样本差异越大
- ### levene检验的原假设是两个样本的方差相同，p值越低,说明两样本差异越大
- ### k-s检验的原假设是两样本分布相同，p值越低,说明两样本差异越大

- ### 按照我们的经验，`市值因子`对市场风格的捕捉能力很强，09年起，小盘股在A股市场一骑绝尘，时而伴随着巨幅的大小盘风格切换。
