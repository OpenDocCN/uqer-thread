```python
__author__ = 'Li'

import pandas as pd
import numpy
import quartz
from quartz.api import *
from matplotlib import pylab

refresh_rate = 10                           # 调仓频率，即每 refresh_rate 个交易日执行一次 handle_data() 函数
universe = set_universe('HS300')  # 证券池，支持股票和基金
start = '2012-12-01'
end = '2015-05-01'
benchmark = 'HS300'
capital_base = 1000000
csvs = []
security_base = {}
commission = Commission(buycost=0.0008, sellcost=0.0018)  # 佣金万八
slippage = Slippage()

lower = 10
higher = 50
testPara = []
out_keys = ['annualized_return', 'volatility', 'max_drawdown']
testPara.append(out_keys)

def initialize(account):                   # 初始化虚拟账户状态
    pass

def handle_data(account):                  # 每个交易日的买入卖出指令
#     print 'lower   higher'
#     print lower, higher
    dt = Date.fromDateTime(account.current_date)

    cal = Calendar('China.SSE')
    lastTDay = cal.advanceDate(dt,'-1B',BizDayConvention.Preceding)
    dt = lastTDay.strftime("%Y%m%d")
    for s in account.universe:
        try:
            pe = DataAPI.MktStockFactorsOneDayGet(tradeDate=dt ,secID=u"",ticker=s[:6],field=u"PE",pandas="1").get('PE')[0]
        except:
            return
        if pe > lower and pe < higher and s not in account.valid_secpos:
            order(s, capital_base / account.referencePrice[s] / 10)
        elif pe > higher and s in account.valid_secpos:
            order_to(s, 0)
    return
def eva():
    
    bt, acct = quartz.backtest(start = start,
                      end = end,
                      benchmark = benchmark,
                      universe = universe,
                      capital_base = capital_base,
                      initialize = initialize,
                      handle_data = handle_data,
                      refresh_rate = refresh_rate)
    perf = quartz.perf_parse(bt, acct)
    tempPara = []
    for k in out_keys:
        tempPara.append(perf[k])
    tempPara.append(lower)
    tempPara.append(higher)
    testPara.append(tempPara)
    print tempPara
    perf['cumulative_returns'].plot()
    perf['benchmark_cumulative_returns'].plot()
    pylab.legend(['current_strategy','HS300'])
    print '-------------------------'

eva()
```
