```python
import pandas as pd
import numpy as np
import itertools
import seaborn as sns
```

```python
# 获取行业每月月末收盘价，相较于上期的变化率
def get_industry_data():
    industry_symbol = DataAPI.IndustryGet(industryVersion=u"SW",industryVersionCD=u"",industryLevel=u"1",isNew=u"1",prntIndustryID=u"",field=u"indexSymbol",pandas="1")['indexSymbol']
    industry_symbol = industry_symbol.tolist()
    data_list = pd.DataFrame()
    for ticker_id in industry_symbol:
        industry_month_data = DataAPI.MktIdxmGet(beginDate=u"20141231",endDate=u"20180120",indexID=u"",ticker=ticker_id,field=u"ticker,endDate,closePrice",pandas="1")
        industry_month_data.sort(['endDate'],ascending=True,inplace=True)
        temp = industry_month_data[['endDate','closePrice']]
        temp.set_index('endDate',inplace=True)
        temp['changeRate'] = temp.pct_change(axis=0) #收益率
        temp.reset_index(inplace=True)

        industry_month_data = pd.merge(industry_month_data, temp, on=['endDate','closePrice'], how='inner')
        industry_month_data.dropna(inplace=True)
        data_list = data_list.append(industry_month_data,ignore_index=True)
    return data_list
```

```python
# 相关系数计算预处理
industry_data = get_industry_data()
industry_data = industry_data.set_index(['endDate','ticker']).unstack()['changeRate']
industry_data_rank = industry_data.rank(axis=1)
industry_data_rank
```

```python
# 相关系数计算
def spearman(item1, item2, df):
    res = pd.DataFrame()
    df1 = df[item1].iloc[0:-1]  #行业代码对应的各个月份的收益率值
    df2 = df[item2].iloc[1:]
    # res = df1.corr(df2,'spearman')
    res = np.corrcoef(df1,df2)
    # print(res[0][1])
    # print("------------")
    return res[0][1]


industry_symbol = DataAPI.IndustryGet(industryVersion=u"SW",industryVersionCD=u"",industryLevel=u"1",isNew=u"1",prntIndustryID=u"",field=u"indexSymbol",pandas="1")['indexSymbol'].tolist()

corr={}
# itertools.product将每每两个行业代码组成一个元组
for x in itertools.product(industry_symbol, repeat=2): # 如(801010,801010)、(801010,801020)
    corr[x] = spearman(x[0], x[1], industry_data_rank) #x[0],x[1]为行业代码，str
#注：corr中的x为元祖，corr值形如{('801880', '801040'): 0.048250255826998607,...}
    
corr = pd.Series(corr) #dict没有hist()方法
corr.hist() 
# corr
```

可以看出行业与行业的一阶滞后的相关性并不高，大部分维持在0附近，我们要找出那些相关系数在统计上是显著的两个行业。相关系数不等于0的显著性检验量为：
$$ t = \frac{r\sqrt{n-2}}{\sqrt{1-r^2}}  ,df = n-2$$
查阅$t$分布表,$t_{(36)0.01} = 2.728$,反推出$|r|=0.43$,即当$|r|>0.43$时,在99%置信水平是统计上显著的。故我们筛选出大于0.43或小于-0.43时的行业相关性。

```python
corr_new = corr[(corr>0.43)|(corr<-0.43)]
print(corr_new)
res = pd.DataFrame(corr_new.sort_values(ascending=0)).reset_index()
res.columns = ['industry1','industry2','corr']
res
```
