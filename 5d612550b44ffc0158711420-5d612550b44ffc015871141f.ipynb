## 期权定价——隐式有限差分法
说明：OptionPrice类实例化时，参数optiontype的值为以下四个之一：'ec','ep','ac','ap'  分别代表欧式看涨期权、欧式看跌期权、美式看涨期权、美式看跌期权

```python
from CAL.PyCAL import *
import numpy as np
import matplotlib.pyplot as plt
from numpy import array
from datetime import datetime
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from scipy.linalg import solve_banded
```

```python
class OptionPrice():
    def __init__(self,spot,strike,T,riskfree,vol,optiontype,M,N):
        self.S=spot
        self.T=T
        self.K=strike
        self.r=riskfree
        self.sigma=vol
        self.type=optiontype
        self.M=M
        self.N=N
        self.z=np.log(self.S)
        self.z_max=self.z+(self.r-0.5*self.sigma**2)*T+4.0*self.sigma*T**0.5
        self.z_min=self.z+(self.r-0.5*self.sigma**2)*T-4.0*self.sigma*T**0.5
        if self.type=='ec':
            self.ic=self.EurCallOption()[0]
            self.bcL=self.EurCallOption()[1]
            self.bcR=self.EurCallOption()[2]
        if self.type=='ep':
            self.ic=self.EurPutOption()[0]
            self.bcL=self.EurPutOption()[1]
            self.bcR=self.EurPutOption()[2]
        if self.type=='ac':
            self.ic=self.AmeCallOption()[0]
            self.bcL=self.AmeCallOption()[1]
            self.bcR=self.AmeCallOption()[2]
        if self.type=='ap':
            self.ic=self.AmePutOption()[0]
            self.bcL=self.AmePutOption()[1]
            self.bcR=self.AmePutOption()[2]
         
        self.dz=(self.z_max-self.z_min)/self.N
        self.dt=1.0*self.T/self.M
        self.a=-0.5*self.sigma**2*self.dt/self.dz**2+0.5*(self.r-0.5*self.sigma**2)*self.dt/self.dz
        self.b=1.0+self.r*self.dt+self.sigma**2*self.dt/self.dz**2
        self.c=-0.5*self.sigma**2*self.dt/self.dz**2-0.5*(self.r-0.5*self.sigma**2)*self.dt/self.dz
        
        self.z_array=np.arange(self.z_min,self.z_max+self.dz,self.dz)
        self.t_array=np.arange(0,self.T+self.dt,self.dt)
        self.Grids=self.cal_Grids()
    def EurCallOption(self):
        ic=lambda x:np.maximum(np.exp(x)-self.K,0.0)
        bcL=lambda x:0
        bcR=lambda x:np.exp(self.z_max)-self.K*np.exp(-(self.T-x))
        return ic,bcL,bcR
    def EurPutOption(self):
        ic=lambda x:np.maximum(self.K-np.exp(x),0.0)
        bcL=lambda x:self.K*np.exp(-(self.T-x))-np.exp(self.z_min)
        bcR=lambda x:0
        return ic,bcL,bcR
    def AmeCallOption(self):
        ic=lambda x:np.maximum(np.exp(x)-self.K,0.0)
        bcL=lambda x:0
        bcR=lambda x:np.exp(self.z_max)-self.K
        return ic,bcL,bcR
    def AmePutOption(self):
        ic=lambda x:np.maximum(self.K-np.exp(x),0.0)
        bcL=lambda x:self.K-np.exp(self.z_min)
        bcR=lambda x:0
        return ic,bcL,bcR

    
    def cal_Grids(self):
        Grids=np.zeros((self.M+1,self.N+1))
        Grids[:,0]=array([self.bcL(i) for i in self.t_array])
        Grids[:,self.N]=array([self.bcR(i) for i in self.t_array])
        Grids[self.M]=array([self.ic(i) for i in self.z_array])
        for i in range(self.M-1,-1,-1):
            b=np.zeros(self.N-1)
            b[0]=Grids[i+1][1]-self.a*Grids[i][0]
            b[self.N-2]=Grids[i+1][self.N-1]-self.c*Grids[i][self.N]
            b[1:self.N-2]=Grids[i+1][2:self.N-1]
            ab=np.zeros((3,self.N-1))
            ab[0]=np.ones(self.N-1)*self.c
            ab[1]=np.ones(self.N-1)*self.b
            ab[2]=np.ones(self.N-1)*self.a
            Grids[i][1:self.N]=solve_banded((1,1),ab,b) # solve_banded()
        return Grids
    
    def draw3d_Grids(self):
        font.set_size(20)
        #t_array=np.linspace(0,0.2,int(0.2/self.dt)+1)
        xx,yy=np.meshgrid(self.z_array,self.t_array)
        # xx,yy=np.meshgrid(self.z_array,t_array)
        figure=plt.figure(figsize=(16,10))
        ax=figure.add_subplot(1,1,1,projection='3d')
        surface=ax.plot_surface(xx,yy,self.Grids,cmap=cm.coolwarm)
        # surface=ax.plot_surface(xx,yy,self.Grids[:int(0.2/self.dt)+1],cmap=cm.coolwarm)
        plt.colorbar(surface,shrink=0.75)
        ax.set_xlabel('$ln(S)$',fontdict={'size':18})
        ax.set_ylabel(r"$\tau$", fontdict={"size":18})
        ax.set_zlabel(u"Option Price", fontdict={"size":18})
        if self.type=='ec':
            ax.set_title(u"隐式差分法计算%s期权价格"%(u"欧式看涨") , fontproperties = font)
        if self.type=='ep':
            ax.set_title(u"隐式差分法计算%s期权价格"%(u"欧式看跌") , fontproperties = font)
        if self.type=='ac':    
            ax.set_title(u"隐式差分法计算%s期权价格"%(u"美式看涨") , fontproperties = font)
        if self.type=='ap':    
            ax.set_title(u"隐式差分法计算%s期权价格"%(u"美式看跌") , fontproperties = font)
```

```python
ec=OptionPrice(100,105,5,0.05,0.2,'ec',100,300)
ep=OptionPrice(100,105,5,0.05,0.2,'ep',100,300)
ac=OptionPrice(100,105,5,0.05,0.2,'ac',100,300)
ap=OptionPrice(100,105,5,0.05,0.2,'ap',100,300)
```
