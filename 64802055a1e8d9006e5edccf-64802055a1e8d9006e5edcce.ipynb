```python
# from quartz_extensions import *
# start = '20230106' # 分析起始时间
# end = '20230606' # 分析结束时间
# benchmark = 'HS300' # 因子分析参考标准
# universe = set_universe('HS300') # 证券池
# refresh_rate = 'month' # 因子分析的调仓频率
# decay_number = 5 # 信号最大衰减周期
# quantile_number = 10 # 分位分组的组数
# function_list = ["winsorize", "neutralize", "standardize"] # 信号处理函数列表，从前到后进行处理
# method = 'simple_long_only' # 组合构建方式
# # ## 自定义因子函数
# # def foo(data, dependencies=['closePrice'], max_window=30):
# #     return data['closePrice'].mean()
 
# ## 注册因子名称
# def initialize(account):
#     # a = Signal('MA30', foo)
#     # account.signal_generator = SignalGenerator(a)
#     pass
    
# ## 调用自定义因子分析函数
# create_factor_tear_sheet('CMRA',
#                         start=start, end=end,
#                         benchmark=benchmark, universe=universe,
#                         refresh_rate=refresh_rate, decay_number=decay_number,
#                         quantile_number=quantile_number, function_list=function_list,
#                         method=method, initialize=initialize)

from quartz_extensions import *
start = '20180601'                                         # 分析起始时间
end = '20201201'                                           # 分析结束时间
benchmark = 'HS300'                                        # 因子分析参考标准
universe = set_universe('A')                           # 证券池
refresh_rate = 'week'                                      # 因子分析的调仓频率
decay_number = 5                                           # 信号最大衰减周期
quantile_number = 5                                        # 分位分组的组数
function_list = ["winsorize", "neutralize", "standardize"] # 信号处理函数列表，从前到后进行处理
method = 'simple_long_only'                                # 组合构建方式

## 自定义因子函数
def foo(data, dependencies=['closePrice'], max_window=30):
    return data['closePrice'].mean()
 
## 注册因子名称
def initialize(account):
    a = Signal('MA30', foo)
    account.signal_generator = SignalGenerator(a)

## 调用自定义因子分析函数    
create_factor_tear_sheet('MA30',
                         start=start, end=end,
                         benchmark=benchmark, universe=universe,
                         refresh_rate=refresh_rate, decay_number=decay_number,
                         quantile_number=quantile_number, function_list=function_list,
                         method=method, initialize=initialize)
```
