# 板块走势

```python
import pandas as pd

begindate="2011-01-01"
enddate="2016-10-01"

datedata=DataAPI.TradeCalGet(exchangeCD=u"XSHG",beginDate=begindate,endDate=enddate,field=['calendarDate','isOpen'])
        #都是说明从1991年开始当天是否交易日
datedata=datedata[datedata['isOpen'] == 1]
date= map(lambda x: x[0:4]+x[5:7]+x[8:10],datedata['calendarDate'].values.tolist())


industrydata = pd.DataFrame(columns=['tmp'],index=date)

industry =   DataAPI.IndustryGet(industryVersion=u"",industryVersionCD=u"010303",industryLevel=u"1",isNew=u"1",field=u"indexSymbol",pandas="1").indexSymbol.tolist()

industryinf = [str(x) for x in industry]

for x in industryinf:
    data = DataAPI.MktIdxdGet(tradeDate=u"",indexID=u"",ticker=x,beginDate=begindate,endDate=enddate,field=u"tradeDate,closeIndex",pandas="1").set_index('tradeDate')
    data.index = map(lambda x: x[0:4]+x[5:7]+x[8:10], data.index.tolist())
    data.columns = [x]
    industrydata = pd.concat([industrydata,data],axis=1)
del industrydata['tmp']

industrydata.plot(figsize= (30,20))
```
