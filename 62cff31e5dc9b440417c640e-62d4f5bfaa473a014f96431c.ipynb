# JOR、SUE因子在可转债中的应用情况

```python
import pandas as pd 
import numpy as np 
import datetime
import scipy.stats as ss
import matplotlib.pylab as plt
```

JOR、SUE都是经典的衡量股票超预期的因子，只不过一个是从技术面角度来衡量，一个是从基本面角度来衡量

因子统一进行正则化、标准化，调仓周期为每月初，不考虑手续费

```python
begin_date = '2019-01-01'
end_date = '2022-06-30'
print('backtest time range',begin_date, end_date)
```

```python
def analysis_pct(pct):
    nav = (1+pct).cumprod()
    maxdrawdown = min((nav - nav.cummax()).div(nav.cummax()).fillna(0))
    annualized_ret = (nav.iloc[-1]-1)/(len(nav)/252.0)
    return '%.2f%%'%(annualized_ret*100), '%.2f'%((annualized_ret-0.03)/(pct.std()*252**0.5)), '%.2f%%'%(-maxdrawdown*100)

def neutralize_factors(factors, date):
    df_stock_industry=DataAPI.EquIndustryGet(
        ticker=factors.index.tolist(),intoDate=date,industryVersionCD=u"010303",field=u"ticker,industryName1").set_index('ticker')
    X_dummy = pd.get_dummies(df_stock_industry.industryName1)
    df_factors = DataAPI.MktStockFactorsOneDayGet(tradeDate=date,ticker=factors.index.tolist(),field=u"ticker,LCAP",pandas="1")
    df_factors.set_index('ticker', inplace=True)
    X = pd.merge(df_factors, X_dummy, left_index=True, right_index=True).dropna()
    idxs = X.index
    y = factors.loc[idxs].values
    X = X.values
    X = np.column_stack([np.ones(len(X)), X])
    R = np.linalg.lstsq(X, y)
    residual = y-X.dot(R[0])
    return pd.Series(residual, index=idxs)

def standardize(series,n=3):
    median = series.quantile(0.5)
    new_median = ((series - median).abs()).quantile(0.50)
    max_range = median + n*new_median
    min_range = median - n*new_median
    series = np.clip(series,min_range,max_range)
    return (series-series.mean())/series.std()
```

一、首先是以zz800为基准，回测观察因子在股票中的效果


```python
beginDate=begin_date
n_groups = 10
index_code_map = {
    'hs300': '000300',
    'zz800': '000906'
}
benchmark = 'zz800'
index_code = index_code_map[benchmark]
b = (datetime.datetime.strptime(beginDate, '%Y-%m-%d') - datetime.timedelta(days=365)).strftime('%Y%m%d')
df_index_details=DataAPI.mIdxCloseWeightGet(ticker=index_code,beginDate=b,field=u"effDate,secShortName,consTickerSymbol,consExchangeCD,weight")
df_index_price = DataAPI.MktIdxdGet(ticker=index_code,beginDate=b,field=u"tradeDate,preCloseIndex,openIndex,lowestIndex,closeIndex")
df_index_price.index = df_index_price.tradeDate
```

```python
def run_stock_backtest(factors_calfuncs, n_groups=n_groups):
    position = pd.Series()
    position_universe = {i:pd.Series() for i in range(n_groups)}
    position_chg = False
    init_month = 0
    tmp_prices = pd.DataFrame()
    tmp_prices_univeres = {i:pd.DataFrame() for i in range(n_groups)}
    df_navs = pd.DataFrame(columns=['benchmark', 'strategy'])
    df_navs = pd.DataFrame(columns=['benchmark']+list(range(n_groups)))
    for i, date in enumerate(df_index_price['tradeDate'][:-1]):
        if date <= beginDate or date >= end_date:
            continue
        month_num = int(date.split('-')[1])
        if month_num != init_month: 
            position_chg = True
            init_month = month_num
        if position_chg: #每月初换仓
            last_date = df_index_price['tradeDate'].iloc[i-1] #换仓前一天作为因子计算的日期
            eff_date = df_index_details['effDate'][df_index_details['effDate']<=last_date].iloc[-1] #记录换仓前的指数成分股有效日期
            stock_codes = df_index_details[df_index_details['effDate']==eff_date]['consTickerSymbol'].sort_values().tolist() #根据日期提取出指数成分股票
            factors = []
            for func in factors_calfuncs:
                factors.append(func(stock_codes, last_date).dropna())
            agg_f = pd.concat(factors, axis=1, join='inner')
            agg_f = agg_f.apply(standardize, axis=0).sum(axis=1) # 根据传入的因子计算方程，计算出单因子数值。如果传入多个因子的计算方程，就把所有的因子标准化之后相加作为最后筛选的总因子
            """
            下面是股票调仓的操作，主要是根据每组之前的总资金，计算调仓后股票的应分配资金
            """
            n = len(agg_f)//n_groups
            for i in range(n_groups):
                top_codes = agg_f.sort_values().iloc[i*n:i*n+n].index.tolist() if i<n_groups-1 else agg_f.sort_values().iloc[i*n:].index.tolist() #提取每组的股票代码
                et = datetime.datetime.strptime(date, '%Y-%m-%d') + datetime.timedelta(days=60)
                et = et.strftime('%Y-%m-%d')
                tmp_prices=DataAPI.MktEqudAdjGet(ticker=top_codes,beginDate=date,endDate=et,field=u"ticker,tradeDate,closePrice,preClosePrice") #提前记录下该组所有股票未来60天的价格数据，用于后续进行每天的净值更新
                tmp_prices = tmp_prices.pivot_table('closePrice', index='tradeDate', columns='ticker')/tmp_prices.pivot_table('preClosePrice', index='tradeDate', columns='ticker') 
                weight = df_index_details[df_index_details['effDate']==eff_date] #根据eff_date那一天的指数的成分股权重，作为调仓之后组内新的股票权重
                weight = weight[weight['consTickerSymbol'].isin(top_codes)][['consTickerSymbol', 'weight']]
                weight = pd.Series(weight['weight'].values, index=weight['consTickerSymbol'])
                weight = weight/weight.sum()
                position = position_universe[i] #position里面的存储的是该组内每只股票的资金
                if position.empty: #如果是策略刚开始建仓，默认初始资金为1，并按照权重进行股票资金分配
                    position = weight
                else:              #后续资金分配需要考虑上一个交易日的组内资金总量
                    position = weight * position.sum()
                position_universe[i] = position #更新调仓后的每组position
                tmp_prices_univeres[i] = tmp_prices  #更新调仓后的组内股票价格信息
            position_chg = False
        df_navs.loc[date, 'benchmark'] = df_index_price.loc[date, 'closeIndex'] #指数价格更新
        for i in range(n_groups): #依据之前记录的股票未来价格信息tmp_prices，每天更新每组内股票的资金规模
            tmp_prices = tmp_prices_univeres[i] #取出该组的股票未来价格信息
            df_pcts = tmp_prices.loc[date]  #提取出当天的股票价格涨跌幅
            position = position_universe[i] #提取出该组的position
            position = position * df_pcts    #按照当天股票的涨跌幅，更新组内每只股票的资金
            assert not pd.isnull(position.sum()) 
            df_navs.loc[date, i] = position.sum() #计算该组当天的总资金，也就是净值
            position_universe[i] = position
    """
    策略运行结束后的净值分析工作，主要是计算收益率、超额收益率、多空对冲收益率等信息
    """
    df_navs = df_navs.pct_change().fillna(0)
    for i in range(n_groups):
        df_navs['alpha'+str(i)] = df_navs[i] - df_navs['benchmark']
    df_navs['top-bottom'] = df_navs[n_groups-1] - df_navs[0]
    df_navs = (1+df_navs).cumprod()
    return df_navs
        
def _plot(df_navs, n_groups=n_groups):
    df_navs[['benchmark']+list(range(n_groups))].plot(title=u'long-only')
    df_navs[['top-bottom']+['alpha'+str(i) for i in range(n_groups)]].plot(title=u'long-benchmark')
    for col in df_navs:
        r1, r2, r3 = analysis_pct(df_navs[col].pct_change())
        print('%s 年化收益、夏普、回撤：%s,%s,%s'%(col, r1, r2, r3))
    
```

1.JOR因子

```python
def calc_jor(stock_codes, date, **kwargs):
    b = (datetime.datetime.strptime(date, '%Y-%m-%d') - datetime.timedelta(days=180)).strftime('%Y-%m-%d')
    d1 = DataAPI.FdmtEeGet(ticker=stock_codes,publishDateBegin=b,publishDateEnd=date,field=u"ticker,publishDate") #业绩快报
    d1['publishDate'] = d1['publishDate'].apply(lambda x: x.split(' ')[0]) 
    d2 = DataAPI.FdmtEfNewGet(ticker=stock_codes,publishDateBegin=b,publishDateEnd=date,field=u"ticker,publishDate")#业绩预告
    d3=DataAPI.FdmtBSGet(ticker=stock_codes,publishDateEnd=date,beginDate=b,field='ticker,publishDate') #定期公告
    df_pubDate = pd.concat([d1,d2,d3])
    df_pubDate = df_pubDate.sort_values('publishDate', ascending=False).drop_duplicates('ticker', keep='first')
    df_pubDate.index=range(len(df_pubDate))
    df1 = DataAPI.MktEqudAdjGet(ticker=stock_codes,beginDate=df_pubDate['publishDate'].min(),endDate=date,field=u"ticker,tradeDate,preClosePrice,lowestPrice")
    df_jors = df1.pivot_table('lowestPrice',index='tradeDate',columns='ticker')/df1.pivot_table('preClosePrice',index='tradeDate',columns='ticker')
    df_stock_index = DataAPI.MktIdxdGet(ticker='000906',beginDate=b,field=u"tradeDate,preCloseIndex,lowestIndex")
    df_stock_index.index=df_stock_index['tradeDate']
    df_jors = df_jors.replace({0:None}).sub(df_stock_index.loc[df_jors.index,'lowestIndex']/df_stock_index.loc[df_jors.index,'preCloseIndex'], axis=0)
    factors = df_pubDate.apply(lambda x: df_jors[df_jors.index>=[x.loc['publishDate']]][x.loc['ticker']].iloc[0],axis=1)
    factors.index  = df_pubDate['ticker'].values
    factors = factors.dropna()
    factors = neutralize_factors(factors, date)
    return factors

df_navs = run_stock_backtest([calc_jor])
_plot(df_navs)
```

2.SUE因子

```python

def calc_sue(stock_codes, date, **kwargs):
    b = (datetime.datetime.strptime(date, '%Y-%m-%d') - datetime.timedelta(days=4*365)).strftime('%Y-%m-%d')
    df_netincomes = DataAPI.FdmtISQPITGet(
        ticker=stock_codes,publishDateEnd=date,beginDate=b,field='ticker,endDate,NIncome').drop_duplicates(['ticker','endDate'],keep='first')
    df_netincomes = df_netincomes.pivot_table('NIncome', index='endDate', columns='ticker')
    df_netincomes.sort_index(ascending=False, inplace=True)
    def calc_sue_impl(net_incomes):
        idx=net_incomes.fillna(method='ffill').dropna().index[0]
        net_incomes = net_incomes.loc[idx:]
        try:
            diff = net_incomes.diff(-4).iloc[1:9].dropna()
            assert len(diff) >= 4
            eps = diff.mean()
            sigma = diff.std()
            factor = (net_incomes.iloc[0] - net_incomes.iloc[4] - eps)/sigma
        except:
            factor = None
        return factor
    factors = df_netincomes.apply(calc_sue_impl, axis=0).dropna()
    factors = neutralize_factors(factors, date)
    return factors

df_navs = run_stock_backtest([calc_sue])
_plot(df_navs)
```

3.JOR+SUE 简单叠加

```python
df_navs = run_stock_backtest([calc_sue, calc_jor])
_plot(df_navs)
```

二、下面尝试把因子用于可转债中进行筛选，观察效果

```python

total_tickers = DataAPI.MktConsBondPremiumGet(beginDate=b,field=u"tickerBond,tickerEqu,SecID")
total_bond_tickers = total_tickers['SecID'].drop_duplicates().tolist()
total_stock_tickers = total_tickers['tickerEqu'].drop_duplicates().tolist()
df_prices = DataAPI.MktConsBondPerfGet(beginDate=b,endDate=end_date,secID=total_bond_tickers,field=u"secID,tickerBond,tradeDate,closePriceBond,tickerEqu,closePriceEqu,debtPuredebtRatio,convPrice,remainSize",pandas="1")
df_bond_mktprices = DataAPI.MktBonddGet(beginDate=b, endDate=end_date,secID=total_bond_tickers,field=u"secID,tradeDate,openPrice,closePrice,turnoverVol")
df_bond_mktprices = df_bond_mktprices.replace({0:None}).fillna(method='ffill')
df_bond_closeprice = df_bond_mktprices.pivot_table('closePrice',index='tradeDate',columns='secID')
df_bond_pcts = (df_bond_closeprice/df_bond_closeprice.shift(1) - 1).fillna(0)
```

```python
df_index_price = DataAPI.MktIdxdGet(ticker=u"000832",beginDate=b,endDate=end_date, field='tradeDate,closeIndex')
df_index_price.index=df_index_price['tradeDate']
df_index_cons = pd.DataFrame()
for day in df_index_price.index:
    df_tmp = DataAPI.IdxConsGet(ticker=u"000832", intoDate=day)
    df_tmp['effDate'] = day
    df_index_cons = df_index_cons.append(df_tmp)
```

```python
def convbonds_factors_fetch(factors_calfuncs, date):
    tmp_prices = df_prices[(df_prices['tradeDate']==date)]    
    bond_ids = df_index_cons[df_index_cons['effDate']==date]['consID'].tolist()
    stock_codes = tmp_prices['tickerEqu'][tmp_prices['secID'].isin(bond_ids)].unique()
    factors = None
    for func in factors_calfuncs:
        if factors is not None:
            factors = factors + standardize(func(stock_codes,date, tmp_prices=tmp_prices))
        else:
            factors = standardize(func(stock_codes,date, tmp_prices=tmp_prices))
    factors = factors.dropna().sort_values()
    return factors,tmp_prices
    
```

```python
def run_bond_backtest(factors_calfuncs,n_groups=n_groups):
    position = pd.Series()
    position_chg = False
    init_month = 0
    fee = 0.000
    position_universe = {i:pd.Series() for i in range(n_groups)}
    df_navs = pd.DataFrame(columns=['benchmark']+list(range(n_groups)))
    for i, date in enumerate(df_index_price['tradeDate'][:-1]):
        if date < begin_date:
            continue
        if date >= end_date:
            break
        month_num = int(date.split('-')[1])
        if month_num != init_month:
            position_chg = True
            init_month = month_num

        if position_chg or init_month==0:
            last_date = df_index_price['tradeDate'].iloc[i-1]
            factors,tmp_prices = convbonds_factors_fetch(factors_calfuncs,last_date)
            n = len(factors)//n_groups
            for i in range(n_groups):
                stock_codes = factors.iloc[i*n:i*n+n].index.tolist() if i<n_groups-1 else factors.iloc[i*n:].index.tolist()                  
                bond_ids = tmp_prices['secID'][tmp_prices['tickerEqu'].isin(stock_codes)].tolist()
                weight = tmp_prices.set_index('secID').loc[bond_ids, 'remainSize']
                weight = weight/weight.sum()
                weight = weight/weight/len(weight)  #这里调仓后的权重为等权分配
                position = position_universe[i]
                if position.empty:
                    position = weight
                else:
                    position = weight * position.sum()
                position_universe[i] = position
            position_chg = False
        df_navs.loc[date, 'benchmark'] = df_index_price.loc[date, 'closeIndex']
        #intraday pcts changed
        """
        这里本意是想把组内债券的收益率计算拆分成日内变化和隔夜变化，以便更精细的计算调仓和手续费的影响
        但实际由于数据质量问题，暂且搁置
        """
        for n in range(n_groups): 
            position = position_universe[n]                   
            df_pcts = df_bond_pcts.loc[date,position.index.tolist()]+1
            position = position * df_pcts
            df_navs.loc[date, n] = position.sum() - fee
            assert not pd.isnull(position.sum()) and not position.empty
            position_universe[n] = position
    df_navs = df_navs.pct_change().fillna(0)
    for i in range(n_groups):
        df_navs['alpha'+str(i)] = df_navs[i] - df_navs['benchmark']
    df_navs['top-bottom'] = df_navs[n_groups-1] - df_navs[0]
    df_navs = (1+df_navs).cumprod()            
    return df_navs
```

三、隐含波动率因子
参考券商报告《华泰证券深度研究：可转债市场概览及其隐含波动率策略》，构建转债特有的隐含波动率因子，进行回测

```python
df_call = DataAPI.BondConvCallTriggerGet(secID=total_bond_tickers,field="secID,exerPar")
df_put = DataAPI.BondConvPutTriggerGet(secID=total_bond_tickers,field=u"secID,ticker,exerPar,startDate,endDate,triggerRate,triggerDays,triggerReachDays",pandas="1")

stock_prices = DataAPI.MktEqudAdjGet(beginDate=b,endDate=end_date,ticker=total_stock_tickers,field='ticker,tradeDate,preClosePrice,closePrice')
df_maturity_day = DataAPI.BondGet(typeID='02020707',field=u"secID,ticker,maturityDate",pandas="1").set_index('secID').drop_duplicates(['maturityDate'])

from scipy import stats
def bsm_value(s,x,t,r,vol):
	d1 = (np.log(s/x)+(r+0.5*(vol**2))*t)/(np.sqrt(t)*vol)
	d2 = d1-vol*np.sqrt(t)
	ret = s*stats.norm.cdf(d1) - x*np.exp(-r*t)*stats.norm.cdf(d2)
	return ret

def bsm_imp_vol(args):
    s,x,t,c = args
    r = 0.02
    c_est = 0
    top=1
    floor=0
    sigma=(top+floor)/2.0
    count=0
    while abs(c - c_est ) > 0.0001:
        c_est = bsm_value(s,x,t,r,sigma)
        count += 1
        if count > 10000 or sigma<0.0001 or sigma >0.9999:
            break
        if c-c_est>0:
            floor=sigma
            sigma=(sigma+top)/2.0
        else:
            top=sigma
            sigma=(sigma+floor)/2.0
    return sigma

def calc_stock_sigma(stock_codes, date):
    b = (datetime.datetime.strptime(date, '%Y-%m-%d') - datetime.timedelta(days=92)).strftime('%Y-%m-%d')
    stock_prices = DataAPI.MktEqudAdjGet(beginDate=b,endDate=date,ticker=stock_codes,field='ticker,tradeDate,preClosePrice,closePrice')
    stock_prices['pcts'] = stock_prices['closePrice']/stock_prices['preClosePrice']-1
    stock_pcts = stock_prices.pivot_table('pcts',index='tradeDate',columns='ticker')
    return stock_pcts.std()*np.sqrt(252)

def calc_option_sigma(bond_ids, date):
    tmp_prices = df_prices[(df_prices['tradeDate']==date)&(df_prices['secID'].isin(bond_ids))].set_index('secID')
    con_prices = tmp_prices['convPrice']
    df_maturity_day = DataAPI.BondGet(secID=bond_ids,field=u"secID,maturityDate",pandas="1").set_index('secID').drop_duplicates(['maturityDate'])
    stock_price =tmp_prices['closePriceEqu']
    bond_prices = tmp_prices['closePriceBond']
    k = con_prices*bond_prices/100.0
    option_value = (tmp_prices['closePriceBond'] - tmp_prices['debtPuredebtRatio'])*con_prices/100.0
    def calc_T(maturity_date,date=date):
        return (datetime.datetime.strptime(maturity_date, '%Y-%m-%d') - datetime.datetime.strptime(date, '%Y-%m-%d')).days/365.0
    T = df_maturity_day.iloc[:,0].apply(lambda x:calc_T(x))
    factors = pd.concat([stock_price,k,T,option_value],axis=1).dropna().apply(lambda x: bsm_imp_vol(x.tolist()), axis=1)
    return factors 

def calc_implied_vol(stock_codes,date, **kwargs):
    tmp_prices = kwargs['tmp_prices']
    bond_ids = tmp_prices[tmp_prices['tickerEqu'].isin(stock_codes)]['secID'].unique().tolist()
    stock_sigmas = calc_stock_sigma(stock_codes, date).dropna()
    option_sigmas = calc_option_sigma(bond_ids, date).dropna()
    option_sigmas.index = [tmp_prices[tmp_prices['secID']==x]['tickerEqu'].iloc[0] for x in option_sigmas.index]
    same_ids = [x for x in stock_sigmas.index if x in option_sigmas.index]
    factors = stock_sigmas.loc[same_ids] - option_sigmas.loc[same_ids]  
    factors = factors[~factors.index.duplicated()]
    return standardize(factors)
```

```python
df_navs = run_bond_backtest([calc_implied_vol],n_groups)
_plot(df_navs,n_groups)
```
