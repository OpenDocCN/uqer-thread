```python
#第五步之现金流：排除近五年现金流净额累计值和净利润加折旧累计值之比小于0.7的股票#这个参数可以调整，我觉得0.8可能比较合适，0.7条件太宽松了
date1 = '20181231'
datelist = ['20171231','20161231','20151231','20141231']
list3 = []
list4 = []
for stock in list6:
    #资产负债表#
    a1 = DataAPI.FdmtBSGet(secID=stock,reportType=u"A",endDate=u"",beginDate=date1,publishDateEnd=u"",publishDateBegin=u"",endDateRep=date1,beginDateRep=date1,beginYear="",endYear="",fiscalPeriod="",field=u"endDateRep,secID,fixedAssets",pandas="1")

    for date in datelist:
        b = DataAPI.FdmtBSGet(secID=stock,reportType=u"A",endDate=u"",beginDate=date,publishDateEnd=u"",publishDateBegin=u"",endDateRep=date,beginDateRep=date,beginYear="",endYear="",fiscalPeriod="",field=u"endDateRep,secID,fixedAssets",pandas="1")
        a1 = a1.append(b)
    #利润表#
    a2 = DataAPI.FdmtISGet(secID=stock,reportType=u"A",endDate=u"",beginDate=date1,publishDateEnd=u"",publishDateBegin=u"",endDateRep=date1,beginDateRep=date1,beginYear="",endYear="",fiscalPeriod="",field=u"endDateRep,secID,NIncomeAttrP",pandas="1")
    for date in datelist:
        b = DataAPI.FdmtISGet(secID=stock,reportType=u"A",endDate=u"",beginDate=date,publishDateEnd=u"",publishDateBegin=u"",endDateRep=date,beginDateRep=date,beginYear="",endYear="",fiscalPeriod="",field=u"endDateRep,secID,NIncomeAttrP",pandas="1")
        a2 = a2.append(b)
    #现金流量表#
    a3 = DataAPI.FdmtCFGet(secID=stock,reportType=u"",endDate=u"",beginDate=date1,publishDateEnd=u"",publishDateBegin=u"",endDateRep=date1,beginDateRep=date1,beginYear="",endYear="",fiscalPeriod="",field=u"endDateRep,secID,NCFOperateA",pandas="1")
    for date in datelist :
        b = DataAPI.FdmtCFGet(secID=stock,reportType=u"",endDate=u"",beginDate=date,publishDateEnd=u"",publishDateBegin=u"",endDateRep=date,beginDateRep=date,beginYear="",endYear="",fiscalPeriod="",field=u"endDateRep,secID,NCFOperateA",pandas="1")
        a3 = a3.append(b)
    Y = sum(a3['NCFOperateA'])/(sum(0.08*a1['fixedAssets']) + sum(a2['NIncomeAttrP']))
    if Y<0.7:
        list3.append(stock)
    elif Y>=0.7:
        list4.append(stock)
print len(list4)
```

```python
#第七步：排除周期股
#计算净利润TTM值
list19 =[]
list110 =[]
T = pd.to_datetime(threeyearsago)
for stock in list4:
    #0、确定计算哪只股票截止到哪天为止的TTM净利润值，记为T日
    #1、获取财报数据，此部分与之前PIT一样处理
    income= DataAPI.FdmtISGet(ticker='',secID=stock, beginDateRep="2012-01-01",field=u"ticker,publishDate,endDate,fiscalPeriod,NIncome",pandas="1")
    income['publishDate'] = pd.to_datetime(income['publishDate'])
    income['endDate'] = pd.to_datetime(income['endDate'])
    #2、只取T日及之前发布的的财报,选取发布的累计值信息
    income = income.loc[income['publishDate'] <= T]
    df = income.loc[income['endDate'].dt.month == income['fiscalPeriod'].astype(int)]
    #3、排序、去重，选取最近一个财报期的财报，记录其财报期为A
    df = df.sort_values(['endDate', 'publishDate'], ascending=[False, False]).drop_duplicates(subset=['endDate'], keep='first')
    latest_date = df.iloc[0].loc['endDate']#有错误#
    #4、选取A，A之前一年的年报记为B，去年与A同期的财报记为C
    last_year_end_date = pd.to_datetime(str(latest_date.year - 1) + '1231')#有没有更好的方法实现
    last_same_season_end_date = latest_date-relativedelta(years=1)
    #5、获取A、B、C中的净利润数据
    df = df.set_index('endDate')
    income_this_season = df.loc[latest_date.date()].loc['NIncome']
    income_last_year = df.loc[last_year_end_date.date()].loc['NIncome']
    income_last_same_season = df.loc[last_same_season_end_date.date()].loc['NIncome']
    #6、截至T日TTM净利润 = A中净利润 + B中净利润 – C中净利润
    income_TTM = income_this_season + income_last_year - income_last_same_season    
    
    #------------------------------------------资产负债表-----------------------------------------#
    #	tradingFA=交易性金融资产，availForSaleFa=可供出售金融资产，cashCEquiv=货币资金
    #   STBorr = 短期借款，LTBorr=长期借款，bondPayable=应付债券，TShEquity=净资产
    a = DataAPI.FdmtBSGet(ticker=u"",secID=stock,reportType=u"",endDate=u"",beginDate=u"",publishDateEnd=u"",publishDateBegin=u"",endDateRep="",beginDateRep="",beginYear="",endYear="",fiscalPeriod="",field=u"secID,endDate,secShortName,tradingFA,availForSaleFa,cashCEquiv,STBorr,LTBorr,bondPayable,TShEquity",pandas="1")
    if np.isnan(a.ix[0,'tradingFA']):
        x1 = 0
    else :
        x1 = a['tradingFA'][0]
    if np.isnan(a.ix[0,'availForSaleFa']):
        x2 = 0
    else :
        x2 = a['availForSaleFa'][0]
    if np.isnan(a.ix[0,'cashCEquiv']):
        x3 = 0
    else :
        x3 = a['cashCEquiv'][0]
    if np.isnan(a.ix[0,'STBorr']):
        y1 = 0
    else :
        y1 = a['STBorr'][0]
    if np.isnan(a.ix[0,'LTBorr']):
        y2 = 0
    else :
        y2 = a['LTBorr'][0]
    if np.isnan(a.ix[0,'bondPayable']):
        y3 = 0
    else :
        y3 = a['bondPayable'][0]
    if np.isnan(a.ix[0,'TShEquity']):
        z1 = 0
    else :
        z1 = a['TShEquity'][0]
    #资本回报率 = 净利润/(净资产+有息负债-现金及其等价物)
    roic = income_TTM/(z1-x1-x2-x3+y1+y2+y3)
    if roic > 0.08:
        list19.append(stock)
    else:
        list110.append(stock)
print len(list19)
print list19
```
