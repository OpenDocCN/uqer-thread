```python
from datetime import datetime,timedelta
```

```python
# secID = '000001.XSHE'
# date = '20160112'
def signal(secID,date):
    beginDate = datetime.strftime(datetime.strptime(date,"%Y%m%d")-timedelta(days = 2),"%Y%m%d")
    df = DataAPI.MktEqudGet(tradeDate=u"",secID=secID,ticker=u"",beginDate=beginDate,
                               endDate=date,isOpen="",field=u"",pandas="1")
    if len(df)==2:
        if df['openPrice'][1] > df['highestPrice'][0] and  df['closePrice'][0]/df['preClosePrice'][0]-1<0.099:
            return "sell"
        elif len(df)==2 and df['openPrice'][1] < df['highestPrice'][0] and df['closePrice'][0]/df['preClosePrice'][0]-1>-0.099:
            return "buy"
    else:
        return "idk"
```
