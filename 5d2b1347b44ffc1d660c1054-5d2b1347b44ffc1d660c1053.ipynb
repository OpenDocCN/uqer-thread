```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import datetime as dt
from CAL.PyCAL import * # CAL provides classes and functions dealing with date, time, and calendars. 

sns.set()
pd.set_option('display.width', 200)
cal = Calendar('China.SSE')
```

```python
secID = '510050.XSHG'
start = '20190101'
end = '20190710'

security = DataAPI.MktFunddGet(secID, beginDate=start, endDate=end, field=['tradeDate', 'closePrice']) # 基金日行情
security['tradeDate'] = pd.to_datetime(security['tradeDate'])
security = security.set_index('tradeDate')
security.info()
security[['closePrice']].plot(figsize=(25,7))
security
```

```python
window_short = 20
window_long = 120
SD = 0.05
security['short_window'] = np.round(pd.rolling_mean(security['closePrice'], window=window_short), 2)
security['long_window'] = np.round(pd.rolling_mean(security['closePrice'], window=window_long), 2)
security[['closePrice', 'short_window', 'long_window']].plot(figsize=(25,7))
security
```

```python
security['s-l'] = security['short_window'] - security['long_window']
security['Regime'] = np.where(security['s-l'] > security['long_window'] * SD, 1, 0)
security.plot(figsize=(25,7))
```
