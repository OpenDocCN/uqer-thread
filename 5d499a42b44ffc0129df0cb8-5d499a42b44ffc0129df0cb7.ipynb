```python
start_date = '20180701' # 开始日期
end_date = (dt.datetime.today() - dt.timedelta(days=1)).strftime('%Y%m%d') # 截止日期

# 交易所交易日历 
"""
XSHG，XSHE，CCFX，XDCE，XSGE，XZCE，XHKG。XSHG表示上海证券交易所，XSHE表示深圳证券交易所，CCFX表示中国金融期货交易所，XDCE表示大连商品交易所，XSGE表示上海期货交易所，XZCE表示郑州商品交易所，XHKG表示香港证券交易所。可同时输入多个证券交易所,可以是列表
"""
TradeCal=DataAPI.TradeCalGet(exchangeCD="XSHG",beginDate=start_date,endDate=end_date,field="calendarDate,isOpen",pandas="1")
TradeCal = TradeCal.loc[TradeCal['isOpen'] == 1,:] # TradeCal = TradeCal[TradeCal['isOpen'] == 1]

Dates = map(lambda x: x[0:4]+x[5:7]+x[8:10], TradeCal['calendarDate'])
PB_PE = pd.DataFrame(index=Dates, data=0, columns=['total PB(LCAP)','total PE(LCAP)','total PB(LFLO)','total PE(LFLO)','closeIndex','stock_value'])
jiaoyi_pinzhongjihe='HS300'
stock_ticker='601398'
for date in Dates:
    """
    上证50(SH50)、上证180(SH180)、沪深300(HS300)、中证500(ZZ500)、创业板(CYB)、中小板(ZXB)、全A股(A)，如"HS300"；另外，支持自定义     指数代码，可以获得date日期的成分股列表，如"000001.ZICN"；对于预设行业实例（IndSW/IndZJH/IndZZ）及预设指数实例（IdxCN），可直接使     用date (datetime.datetime): 交易日
    """
    universe = set_universe(jiaoyi_pinzhongjihe, date) # + set_universe('ZZ500', date)
    
    # 获取多只股票历史上某一天的因子数据
    data = DataAPI.MktStockFactorsOneDayGet(tradeDate=date,secID=universe[:],field=['ticker','LCAP','LFLO','PB','PE'],pandas="1").set_index('ticker')
    """
    LCAP	float	对数市值（Natural logarithm of total market values）。计算方法：对数市值=市值的对数。属于价值类因子。数据更新     时间：T日21点30分。
    LFLO	float	对数流通市值（Natural logarithm of float market values）。计算方法：对数流通市值=流通市值的对数。属于价值类因     子。数据更新时间：T日21点30分。
    """
    
    data_total = data.shape[0]
    data.dropna(inplace=True)
    # 数据量不足80%时，不具有代表性
    if float(data.shape[0]) / data_total < 0.8:
        continue
    
     # 细节处理：负值极值处理
    data['PB'][data['PB'] < 0] = 0
    data['PE'][data['PE'] < 0] = 0
    pb = winsorize(data['PB'].to_dict())
    pe = winsorize(data['PE'].to_dict())
    data['PB'][pb.keys()] = pb.values()
    data['PE'][pe.keys()] = pe.values()
    
    # 权重
    data['LCAP'] = data['LCAP'] / sum(data['LCAP'])
    data['LFLO'] = data['LFLO'] / sum(data['LFLO'])
    
    # 指数日行情
    MktIdxdGet_000001=DataAPI.MktIdxdGet(indexID='000001.ZICN',tradeDate=date,field='closeIndex',pandas="0")
    MktIdxdGet_000001=float(MktIdxdGet_000001[11:20])
    
    # 选择的股票的行情
    stock_value=DataAPI.MktEqudGet(ticker="000001",tradeDate=date,field="closePrice",pandas="0")
    stock_value=float(stock_value[11:16])
    
    PB_PE.loc[date] = [sum(data['LCAP'] * data['PB']), sum(data['LCAP'] * data['PE']),sum(data['LFLO'] * data['PB']), sum(data['LFLO'] * data['PE']),MktIdxdGet_000001,stock_value]

# 增加4列
PB_PE['average PB(LCAP)'] = PB_PE['total PB(LCAP)'].mean()
PB_PE['average PE(LCAP)'] = PB_PE['total PE(LCAP)'].mean()
PB_PE['average PB(LFLO)'] = PB_PE['total PB(LFLO)'].mean()
PB_PE['average PE(LFLO)'] = PB_PE['total PE(LFLO)'].mean()

# 画图
PB_PE[['total PB(LCAP)','average PB(LCAP)','closeIndex']].plot(secondary_y=['closeIndex'],figsize=(16,5))
PB_PE[['total PB(LCAP)','average PB(LCAP)','stock_value']].plot(secondary_y=['stock_value'],figsize=(16,5))

PB_PE[['total PE(LCAP)','average PE(LCAP)','closeIndex']].plot(secondary_y=['closeIndex'],figsize=(16,5))
PB_PE[['total PE(LCAP)','average PE(LCAP)','stock_value']].plot(secondary_y=['stock_value'],figsize=(16,5))

PB_PE[['total PB(LFLO)','average PB(LFLO)','closeIndex']].plot(secondary_y=['closeIndex'],figsize=(16,5))
PB_PE[['total PB(LFLO)','average PB(LFLO)','stock_value']].plot(secondary_y=['stock_value'],figsize=(16,5))

PB_PE[['total PE(LFLO)','average PE(LFLO)','closeIndex']].plot(secondary_y=['closeIndex'],figsize=(16,5))
PB_PE[['total PE(LFLO)','average PE(LFLO)','stock_value']].plot(secondary_y=['stock_value'],figsize=(16,5))

PB_PE
```
