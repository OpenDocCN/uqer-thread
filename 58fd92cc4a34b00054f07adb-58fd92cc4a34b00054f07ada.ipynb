```python
import numpy as np
import pandas as pd
from sklearn import linear_model

Tdate="20170420"
T1date='20170421'

StkDay='20160831'    #股票交易集的日期
```

以下是X：

```python
#获取所有股票之后再获取所在行业
IndSTKbeforeNumeric = DataAPI.EquIndustryGet(industryVersionCD=u"010303",secID=DataAPI.EquGet(secID=set_universe('A'),listStatusCD=u"L",field=u"",pandas="1")['secID'].tolist(),intoDate=StkDay,field=['secID','industryName1'],pandas="1")

#改列名
IndSTKbeforeNumeric.columns = ['secID','Ind']
IndSTK = IndSTKbeforeNumeric

#提取行业列表
IndSTKbeforeNumeric2 = IndSTKbeforeNumeric.groupby(IndSTKbeforeNumeric['Ind'])
INDSTList = IndSTKbeforeNumeric2.any().index.values.tolist()

#提取dummies值
IndSTKdummies = pd.concat([IndSTK,pd.get_dummies(IndSTK['Ind'])],axis = 1)

#获取因子值求倒数并标准化
factor_pfm = DataAPI.MktStockFactorsOneDayGet(tradeDate=Tdate,secID=IndSTKdummies['secID'],field=["secID"]+['PB'],pandas="1")
IndSTKDummiesFactor = pd.merge(IndSTKdummies,factor_pfm)
IndSTKDummiesFactor.insert(0,'BP',1.0/IndSTKDummiesFactor['PB'])
IndSTKDummiesFactor.insert(0,'Stderized_BP',(IndSTKDummiesFactor['BP']-pd.np.mean(IndSTKDummiesFactor['BP']))/pd.np.std(IndSTKDummiesFactor['BP']))

#删除列
del IndSTKDummiesFactor['BP']
del IndSTKDummiesFactor['PB']
del IndSTKDummiesFactor['Ind']
del IndSTKDummiesFactor['secID']
```

以下是Y：

```python
#根据股票列表获取T+1日的pct
ClosePrices = DataAPI.MktEqudAdjGet(tradeDate=T1date,secID=IndSTK['secID'],field=u"secID,closePrice,preClosePrice",pandas="1")#.set_index('secID')
pct_t1 = ClosePrices['closePrice']/ClosePrices['preClosePrice']-1
IndSTKDummiesFactor.insert(0,'pct_t1',pct_t1)
```
