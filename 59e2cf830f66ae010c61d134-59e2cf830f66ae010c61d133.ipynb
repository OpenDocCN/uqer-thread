pandas相关笔记

1、排序

data.sort_values(by=['group','ounces'], ascending=True, inplace=True)  #多于一列的排序，group降序，ounces升序，原位排序；默认排序是升序

2、切片/索引

df.loc[:,['A','B']]   #.loc 是按照标签切片，既包前也包后
df.iloc[3:5,0:2]    #.iloc 是按照位置切片，包前不包后
df.ix[3:5,0:2]      #.ix 混合索引，既可以按照标签，也可以按照位置，当二者相同时，优先按照标签索引
df[df.A > -0.5]   # 布尔索引，返回A列大于-0.5的所有值：先判断A列值是否大于-0.5
df2[df2['E'].isin(['two','four'])]  # read as "return rows where column E contains two or four"
df2[df2.E.str.contains("tw|ou")]  #E列中包含tw或者ou


3、删除/新建

data.drop_duplicates( )   #删除一模一样的行
data.drop_duplicates(subset='k1',keep='last')  #subset按照k1去重，keep保留重复中最后一个
data.drop('animal', axis='columns', inplace=True)  # 删除一列


4、apply/lambda/map

data.animal.apply(lambda x:x[:2])        #取animal列前两位,平时用map多一点，很容易获取数据，apply避免循环
data.animal.apply(lambda x:x+'_good')     #apply可以批量做


5、分组group

df.groupby('key2').max()  #groupby:按key2列分组求最大值;首先将df分成两组，然后在分组内分别进行计算，最后再合在一起——>应用：求不同行业PE最大值
grouped = df['data1'].groupby(df['key1'])  #按key1进行分组，并返回data1列的数据
df['key3'].groupby(df['key1'], df['key2'])  #先根据key1、再根据key2对key3列进行分组
df.groupby('key1').head(1)  #获取每个分组第一行
df.groupby('key1').apply(lambda x:x.sort_values(by='data1',ascending=0).head(1))  #按照key1做分组，再按照data1从大到小排序，再取各组第一个数据——>即取各分组最大的数据


6、替换值replace

data.replace(-999,np.nan,  inplace=True)   #替换值，inplace=True是原位替换
data.replace(np.nan,data.mean(),inplace=True)  #以平均值替换缺失值
data.replace([-999, -1000], [np.nan,100], inplace=True)  #同时替换多个值，inplace

7、缺失值数据处理

df.fillna(0) #用0填充缺失值 df.fillna('missing') 用字符串代替缺失值
df.fillna(method='pad')#用前一个数据代替NaN
df.fillna(method='ffill')  #ffill从上往下添加，添加值为前一个数
df.fillna(method='bfill',limit=1)#用后一个数据替代NaN,限制每列只能替代一个NaN
df.fillna('不知道')  #以具体数值填充缺失值
df.dropna(axis=0)   #删除含有NaN的行，axis=1 删除列
