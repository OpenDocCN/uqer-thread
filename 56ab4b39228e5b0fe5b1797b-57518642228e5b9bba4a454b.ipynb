# 凸优化(Convex Optimization)介绍！

在金融学和经济学中，凸优化(Convex Optimization)起着重要的作用。如有约束优化可解决投资于两种高风险证券的投资者的[效用最大化](http://wiki.mbalib.com/zh-tw/%E6%95%88%E7%94%A8%E6%9C%80%E5%A4%A7%E5%8C%96)问题，或是根据市场数据校准期权定价模型等。

1. 本文参考《[Python.for.Finance](http://pan.baidu.com/s/1eRi5MkM)》
2. 使用 Scipy 与 Numpy 实现。


接下来以对fm函数进行优化为例：

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import matplotlib as mpl
mpl.rcParams['axes.unicode_minus'] = False # 解决保存图像是负号'-'显示为方块的问题
```

```python
def fm((x,y)):
    t = (np.sin(x) + 0.05* x ** 2
         + np.sin(y) + 0.05* y ** 2)
    return t
```

```python
x = np.linspace(-10, 10, 50)
y = np.linspace(-10, 10, 50)
x, y = np.meshgrid(x, y)
z = fm((x, y))
```

下面画出上述函数在定义的x和y区间内的图形。

```python
fig = plt.figure(figsize=(9,6))
ax = fig.gca(projection='3d')
surf = ax.plot_surface(x, y, z, rstride=2, cstride=2,
                      cmap=mpl.cm.coolwarm, linewidth=0.5,
                      antialiased=True)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('f(x, y)')
fig.colorbar(surf, shrink=0.5, aspect=5)
```

从视觉上可以看出这个函数由多个局部极小值。在这个特定的图形表现中无法真正确认全局最小值是否存在。  
接下来，我们打算实施全局最小化方法和局部最小化方法。所使用的 brute 和 fmin 函数可以在子库 scipy.optimize 中找到：  

```python
import scipy.optimize as spo
```

## 一、全局优化(Global Optimization)
---
为了更仔细的观察最小化过程时后台发生的情况，我们改善原函数，通过一个选项输出当前参数值和函数值：

```python
def fo((x,y)):
    z = (np.sin(x) + 0.05*x**2
         + np.sin(y) + 0.05*y**2)
    if output == True:
        print '%8.4f %8.4f %8.4f' %(x, y, z)
    return z
```

正如后面的代码和对应的输出所示，这使我们可以跟踪过程的所有相关信息。brute以参数范围作为输入。例如，提供x值范围(-10, 10.1, 5)会得到-10、-5、 0、5、 10这5个“测试值”：

```python
#brute的用法如下：
spo.brute?
```

```python
output = True
spo.brute(fo, ((-10,10.1,5), (-10,10.1,5)), finish=None)
```

在给定的函数初始参数化条件下，最优化参数值是x=y=0。回顾前面的输出可以看出，结果函数值也为0.第一次参数化相当粗糙，我们对两个输入参数均使用5的步距。这当然有很大的微调空间，得到更好的结果：

```python
output = False
opt1 = spo.brute(fo,((-10,10.1,0.1), (-10,10.1,0.1)), finish=None)
opt1
```

```python
fm(opt1)
```

最优化参数值为x = y = -1.4，全局最小化的最小函数值大约为 -1.7749  
## 二、局部优化(Local Optimization)
---
对于局部凸优化，我们使用全局优化的结果。fmin函数的输入是需要最小化的函数和其实参数值。此外，可以定义输入参数宽容度和函数值宽容度，以及最大迭代及函数调用次数：

```python
#fmin的用法如下：
spo.fmin?
```

```python
output = True
opt2 = spo.fmin(fo, opt1, xtol=0.001, ftol=0.001, maxiter=15, maxfun=20)
opt2
```

同样，我们可以观察到解决方案的微调和更低的函数值：

```python
fm(opt2)
```

在许多凸优化问题中，建议在局部优化之前进行全局优化。主要原因是局部凸优化算法很容易陷入某个局部最小值，而忽略“更好”的局部最小值和全局最小值。下面可以看到，将初始参数化设置为x = y = 2 得出大于0的“最小值”：

```python
output = False
spo.fmin(fo, (2.0, 2.0), maxiter=250)
```

## 三、有约束优化(Constrained Optimization)
---
之前只考虑了无约束优化问题，但是许多类型的经济学或者金融学优化问题都有一个或多个约束条件。这些约束可能采取等式或者不等式等正规形式。  

举个简单的例子，考虑可能投资于两种高风险证券(期望[效用最大化](http://wiki.mbalib.com/zh-tw/%E6%95%88%E7%94%A8%E6%9C%80%E5%A4%A7%E5%8C%96))的投资者的效用最大化问题。两种证券今天的价格为$ q_{a}=q_{b}=10 $。一年之后，状态u下他们的收益分别为 15 和 5，而在状态d下收益分别为 5 和 12。两种状态的出现可能性相同，两种证券的向量收益分别记为$ r_{a} 和 r_{b} $。  

投资者的投资预算为$w_{0}=100$, 并根据效用函数$u(w) = \sqrt{w}$得出未来的财富效率，其中w是可用的财富。  

下式(1)是最优化问题的公式，其中a，b是投资者购买的证券数量:  

$$
\begin{matrix}
\underset{a,b}{max} - E(u(w_{1})) = p\sqrt{w_{1u}}+(1-p)\sqrt{w_{1d}}\\ 
w_{1}=ar_{a} + br_{b}\\ 
w_{0}=aq_{a} + bq_{b}\\ 
a,d\geqslant 0
\end{matrix}
{(1)}
$$

代入所有数值化假设，就得到下式(2)中的问题。与此同时，我们还为负数的预期效用最小化做了更改。  

$$
\begin{matrix}
\underset{a,b}{min} - E(u(w_{1})) = (0.5\sqrt{w_{1u}}+0.5\sqrt{w_{1d}})\\ 
w_{1w}=a\cdot 15 + b\cdot 15\\ 
w_{1d}=a\cdot 5 + b\cdot 12\\ 
100\geqslant a\cdot 10 + b\cdot 10\\ 
a,d\geqslant 0
\end{matrix}
 {(2)}
$$

我们使用 scipy.optimize.minimize 函数解决上述问题。除了需要最小化的函数之外，这个函数还以公式、不等式（以字典对象列表形式）和参数范围（以元祖兑对象形式）作为输入。我们可以将上式(2)中的问题翻译为如下代码：

```python
# function to be minimized
from math import sqrt
def Eu((s, b)):
    t = -(0.5*sqrt(s*15 + b*5) + 0.5*sqrt(s*5 + b*12))
    return t

# constraints
cons = ({'type': 'ineq','fun': lambda (s, b): 100 - s * 10 - b * 10})

# budget constraint
bnds = ((0,1000), (0,1000))
```

现在，我们已经拥有了使用minimize函数所需的一切——只需要添加对最优参数的初始猜测：

```python
result = spo.minimize(Eu, [5,5], method='SLSQP', bounds=bnds, constraints=cons)
result
```

该函数返回一个字典对象。最优参数可以如此读出：

```python
result['x']
```

最优函数值如下（同样改变符号）：

```python
-result['fun']
```

按照对上述简单模型的参数化，投资者的最优选择是购买8个单位的a证券和大约2个单位的b证券。预算是固定，也就是说，投资者在证券在总共投入100资金。这很容易通过去的最优参数向量和价格向量的点乘(dot product)来验证：  
两种证券今天的价格为$ q_{a}=q_{b}=10 $。

```python
np.dot(result['x'], [10, 10])
```
