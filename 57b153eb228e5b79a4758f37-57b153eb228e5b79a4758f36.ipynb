#### 一个简单的市场监视器，可以按照自己喜欢的刷新频率及输出频率输出结果

```python
def market_distribution_monitor(checktime, refresh_rate=60, detail=True):
    """输出目前市场所有股票的涨跌分布"""
    
    import time
    import datetime
    import numpy as np
    import matplotlib.pyplot as plt

    is_first = 1
    while True:
        if is_first or datetime.datetime.now().strftime("%H:%M") in checktime:
            # 第一次输出，如果设成1则每次运行都会输出
            is_first = 0

            dist = DataAPI.MktTickRTSnapshotGet(exchangeCD='XSHG,XSHE',assetClass='E', field='changePct')['changePct']
            bins = np.arange(-0.1, 0.11, 0.01)
            nsec = plt.hist(dist, bins=bins)[0]

            print datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
            if detail:
                for i, n in enumerate(bins[:-1]):
                    print "{:.0f}% - {:.0f}%: {}".format(-bins[i+1]*100, -n*100, int(nsec[-i-1]))
            plt.show()
            print ""

        # 刷新频率
        time.sleep(refresh_rate)
```

```python
checktime = ['09:35', '10:00', '10:30', '11:00', '11:20', '13:10', '13:30', '14:00', '15:00']

market_distribution_monitor(checktime)
```
