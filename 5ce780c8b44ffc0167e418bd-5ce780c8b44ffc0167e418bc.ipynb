```python
a=DataAPI.EquGet(secID=u"",ticker=u"",equTypeCD=u"A",listStatusCD=u"",field=u"",pandas="1")
IndexCode=a.ticker 
#SWIndexFirstClass=SWIndex[SWIndex.ticker.isin(pd.Series(a.ticker)) ]
#SWIndexFirstClass.reset_index(drop=True)

a=a[ ~ a['listStatusCD'].str.contains('DE')]  #删除某列包含特殊字符的行
a=a[ ~ a['listStatusCD'].str.contains('UN')] 
a=a[ ~ a['listStatusCD'].str.contains('S')] 
a=a[ ~ a['secShortName'].str.contains('ST')] 
```

```python
## 获得日行情。
beginDate='20100601'
endDate='20110601'
indexData1=DataAPI.MktEqudGet(beginDate=beginDate,endDate=endDate,secID=u"",ticker=IndexCode,field=u"",pandas="1")[['secID','tradeDate','closePrice','chgPct','PE','marketValue','isOpen']]
indexData1[(True-indexData1['isOpen'].isin([0]))]
## 求得每个交易日的收益率。从而求得每个月的波动率。
#indexData['tradeMonth']=indexData.tradeDate.apply(lambda x:x[:7])
b=pd.DataFrame()
for i in IndexCode:
    b=b.append(DataAPI.MktStockFactorsDateRangeGet(secID=u"",ticker=i,beginDate=beginDate,endDate=endDate,field=u"",pandas="1")[['secID','tradeDate','HBETA']])

indexData = pd.merge(indexData1, b , on=['secID', 'tradeDate'])
indexData['avgPct']=(indexData.chgPct*indexData.marketValue)
index=indexData.groupby(['tradeDate','secID'])['chgPct','avgPct','PE','HBETA','marketValue'].mean()
```

```python
def rank_return(x):
    ## 市值越大，排名值越大

    #x['marketValueRank']=x['marketValue'].rank()
    ## PE值越小，排名值越大
   
    #x['PERank']=x['PE'].rank(ascending=False)
    ## Beta值越小，排名值越大
   
    x['HBETARank']=x['HBETA'].rank(ascending=False)
    return x
    pass

def rank_finalScore(x):
    ## score值越小，排名值越小
    ## score要选【0-6】的
    x['finalRank']=x['score'].rank(ascending=True)
    return x
    pass

rankRtn=index.groupby(level=[0]).apply(rank_return)

```

```python
rankRtn['score']=rankRtn.HBETARank
finalRank=rankRtn.groupby(level=[0]).apply(rank_finalScore)
finalRank = finalRank.dropna(subset=["HBETA"])
```

```python
highrisk=finalRank
t1=0.1
highrisk=highrisk[highrisk['marketValue']<=30000000000]
highrisk=highrisk[highrisk['PE']>=50]
highrisk=highrisk[highrisk['HBETARank']<=3150*t1]
highrisk.to_excel('highrisk.xlsx')
```
