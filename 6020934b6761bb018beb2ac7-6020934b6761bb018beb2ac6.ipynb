```python
begin_date, end_date = '20180101', '20210207'
dates = ['20180701','20190101','20190701','20200101','20200701','20210207']
res = set(set_universe('SH50' ,date='20180101'))
for i in dates:
    res = res.intersection(set(set_universe('SH50' ,date=i)))
res = DataAPI.EquGet(secID=res,ticker=u"",equTypeCD=u"A",listStatusCD=u"",exchangeCD="",ListSectorCD=u"",field=["secID","secShortName"],pandas="1")
res
```

```python
names = ['s1','s2','s3','s4','s5','s6','s7','s8','s9','s10']
repStocks =['601628.XSHG','600016.XSHG', '600309.XSHG', '601336.XSHG', '600104.XSHG', '601601.XSHG', '601688.XSHG', '601166.XSHG', '600887.XSHG', '600519.XSHG']
repStocks

# get price
price = DataAPI.MktEqudGet(secID='600000.XSHG',beginDate=begin_date,endDate=end_date,field=u"closePrice",pandas="1").values
stock_price = pd.DataFrame(price,columns=['s1'])
for i,j in zip(repStocks[1:],names[1:]):
    stock_price[j] = DataAPI.MktEqudGet(secID=i,ticker=u"",tradeDate=u"",beginDate=begin_date,endDate=end_date,isOpen="",field=u"closePrice",pandas="1").values
etf50_close = DataAPI.MktIdxdGet(indexID=u"000016.ZICN",beginDate=begin_date,endDate=end_date,field=u"closeIndex",pandas="1").values

# OLS
y = etf50_close
X = stock_price
X = sm.add_constant(X)
model = sm.OLS(y, X)
results = model.fit()
results.rsquared
```
