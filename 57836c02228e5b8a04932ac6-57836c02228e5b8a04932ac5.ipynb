# 一.前言
在股票组合构建的过程中，我们常常会遇到选行业，选个股，择时这三个主要的问题，三者究竟孰轻孰重，何时择时最重要，哪些行业需要精选个股，哪些行业不需要强调选股，我们该如何抉择取舍呢？“20100325-广发证券-量化投资专题报告：从判定系数R^2的角度看市场”一文提供了一种新的角度来看待这个问题。

1.文章通过市场指数,行业超额收益相对个股收益做两阶段线性回归，得到两类R^2值。

2.以市场全部A股一步线性回归模型R^2的中位数构成股指效应指数，该指标越大，则表明个股受市场影响大，此时可更关心择时；该指标下降，则表明股指效应对股票收益贡献降低，由下往上精选个股将产生超额收益。

3.以市场回归方程中残差相对行业超额收益进行二步回归中的R^2中位数构成行业效应指数，该指数越大，则表明精选行业进行配置能提高超额收益；若行业效应长期偏低，则应把握行业内部公司的差异，精选个股贡献收益

# 二.求R^2
下面来通过优矿平台来便捷地实现R^2的求解过程,先确立一下需要做的两个回归：
\begin{align*}
R(t) = \alpha_1+\beta_1*R_M(t)+e_1(t).....(1)
\end{align*}
\begin{align*}
e_1(t)= \alpha_2+\beta_2*(R_{industry}(t)-R_M(t))+e_2(t).....(2)
\end{align*}

## 1.股指效应指标
下面我们着手开始第一个回归，看一下较近一段时间内股指效应的情况：

研究时间：20140630-20160630

回归样本长度：120个交易日的数据（也就是说研究时间的前120个交易日是没有结果的）

研究股票范围：全A股

市场指数：上证综指

逻辑：每日，每只股票前120个交易日的收益率序列(响应变量)都要对股指收益率序列(解释变量)进行回归，取这两千多个回归R^2的中位数作为当日的股指效应指标

```python
import pandas as pd
import numpy as np

#取交易日
from CAL.PyCAL import *
datedata=DataAPI.TradeCalGet(exchangeCD=u"XSHG",beginDate='20140630',endDate='20160630',field=['calendarDate','isOpen']) 
datedata=datedata[datedata['isOpen'] == 1]
date = map(lambda x: x[0:4]+x[5:7]+x[8:10], datedata['calendarDate'].values.tolist())   

#取所有A股
stocks = DataAPI.EquGet(equTypeCD='A',listStatusCD='L',field='secID,nonrestfloatA')
universe = stocks['secID'].tolist()
```

```python
#算上证综指的每日收益序列
Pm=DataAPI.MktIdxdGet(ticker=u"000001",beginDate=u"20140630",endDate=u"20160630",field=u"preCloseIndex,closeIndex")
Rm=Pm['closeIndex']/Pm['preCloseIndex']-1
```

```python
#取所有A股的每日收益序列，以ticker为列名
R=pd.DataFrame()
for i in universe:
    P=DataAPI.MktEqudGet(secID=i,beginDate=u"20140630",endDate=u"20160630",field=u"preClosePrice,closePrice").dropna() 
    R[i]=P['closePrice']/P['preClosePrice']-1
```

```python
R.shape
```

可以看到一共是491个交易日，和2870只股票，但其中有几只是最近才上市的，之前没有数据，待会儿运行的时候会解决这个问题

下面开始做回归:(PS:这里要运行很长时间，感兴趣的可以直接用我跑出来的结果)

```python
#每日作回归时，取前面120个交易日的数据作回归
from sklearn import linear_model
import statsmodels.api as sm
R2inf=pd.DataFrame()  #用于存储最后得到的每日R^2中位数
R2inf['R2']=np.zeros(372)
R2inf.index=date[119:]   #以交易日为索引
for i in range(119,len(date)):  #对交易日做循环
    R2tmp=np.zeros(2870)   #用于存储当日每只个股R2
    for j in range(2870):
        try:
            X=np.zeros((1,120))
            X[0]=Rm[(i-119):(i+1)]    #解释变量为市场指数前120个交易日的收益序列
            X=X.T
            if np.isnan(X).any():
                continue
            X = sm.add_constant(X,has_constant='skip')
            y=np.array(R.iloc[(i-119):(i+1),[j]])   #响应变量为个股前120个交易日的收益序列
            if np.isnan(y).any():
                continue
            model = sm.OLS(y,X)
            results = model.fit()
            R2tmp[j]= results.rsquared          #存储个股的R^2到数组里
        except:
            continue
    R2tmp=R2tmp[R2tmp>0]                  #排除异常没有赋值的数据
    R2inf.iat[i-119,0] = np.percentile(R2tmp,50)  #取所有个股回归R^2的中位数
```

```python
R2inf.to_csv('R2inf.csv')   #存储起来，免得以后再跑
```

从全A股的情况看，在15年大盘跌之前，股指效应指数长期偏低，在这段时间里，我们可以通过精选个股来提高收益；15年6月大盘跌以后，股指效应指数持续上升，说明股票受大盘影响越来越大，这时候我们可以把重心放在择时上，也可以按照降低非系统性风险的逻辑来精选个股，选取R2高的个股或许能获取超额收益。

为了算其它行业方便，干脆把这个回归的过程定义成一个函数或者一个包：

```python
from sklearn import linear_model
import statsmodels.api as sm
def getR2inf(indexSymbol,industryID):
    #算输入行业指数的每日收益序列
    Pm0=DataAPI.MktIdxdGet(ticker=str(indexSymbol),beginDate=u"20140630",endDate=u"20160630",field=u"preCloseIndex,closeIndex")
    Rm0=Pm0['closeIndex']/Pm0['preCloseIndex']-1
    #取所有该行业的股票的每日收益序列，以ticker为列名
    R0=pd.DataFrame()     
    for i in dict0[industryID]:
        if i != dict0[industryinf.iat[21,0]][0]:
            P=DataAPI.MktEqudAdjGet(secID=i,beginDate=u"20140630",endDate=u"20160630",field=u"preClosePrice,closePrice").dropna() #前复权价格
            R0[i]=P['closePrice']/P['preClosePrice']-1
    v2=Rm0-R_m    #第二步回归的解释变量行业指数的超额收益
    #去除R050里那几个异常交易日（没有行业指数）的数据
    R_0=R0
    for i in b:
        R_0=R_0.drop(i)
    R_0.reset_index(inplace=True)  
    R2inf0=pd.DataFrame()   
    R2inf0['MarketR2']=np.zeros(362)
    R2inf0['IndustryR2']=np.zeros(362)
    R2inf0.index=date2[119:]
    for i in range(119,len(date2)):   #对交易日作循环
        R2tmp1=np.zeros(R_0.shape[1])  #用于存储第一步回归的R^2
        R2tmp2=np.zeros(R_0.shape[1])   #用于存储第二步回归的R^2
        for j in range(R_0.shape[1]):
            X=np.zeros((1,120))
            X[0]=R_m[(i-119):(i+1)]  #第一步回归的解释变量，股指120个交易日的收益率序列
            X=X.T
            try:
                if np.isnan(X).any():
                    continue
                X = sm.add_constant(X,has_constant='skip')
                y=np.array(R_0.iloc[(i-119):(i+1),[j+1]])   #第一步回归的响应变量，个股120个交易日的收益率序列
                if np.isnan(y).any():
                    continue
                model = sm.OLS(y,X)
                results = model.fit()
                R2tmp1[j]= results.rsquared    #第一步回归的R^2
                e= results.resid           #第一步回归的残差序列，也是第二步回归的响应变量
                X=np.zeros((1,120))
                X[0]=v2[(i-119):(i+1)]      #第二步回归的解释变量，行业指数收益减股指收益
                X=X.T
                if np.isnan(X).any():
                    continue
                X = sm.add_constant(X,has_constant='skip')
                model = sm.OLS(e,X)
                result2 = model.fit()
                R2tmp2[j]= result2.rsquared    #第二步回归的R^2
            except:
                continue
        R2tmp1=R2tmp1[R2tmp1>0]
        R2tmp2=R2tmp2[R2tmp2>0]
        R2inf0.iat[i-119,0] = np.percentile(R2tmp1,50)
        R2inf0.iat[i-119,1] = np.percentile(R2tmp2,50)   #取R^2的中位数
    return R2inf0
```
