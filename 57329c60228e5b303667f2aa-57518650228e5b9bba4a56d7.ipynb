# H_A比较研究

```python
import pandas as pd
import numpy as np


#设置研究股票
secID_a = '601398.XSHG'
secID_h= '01398.XHKG'

#设置计算周期
start = '20060104'
end = ''


r = DataAPI.MktFxRefRateGet(currencyPair=u"HKD/CNY",beginDate= start,endDate = end,field=u"",pandas="1")
a = DataAPI.MktEqudGet(tradeDate=u"",secID= secID_a,ticker=u"",beginDate= start,endDate = end,isOpen="",field=u"secID,tradeDate,closePrice",pandas="1")
h = DataAPI.MktHKEqudGet(secID=secID_h,ticker=u"",tradeDate=u"",beginDate= start,endDate = end,field=u"secID,tradeDate,closePrice",pandas="1")

s_temp = pd.merge(a,h,on = 'tradeDate')
security = pd.merge(s_temp,r,on = 'tradeDate')

security['h_a'] = security['closePrice_y']*security['midRate']/security['closePrice_x']

security['v_m']=security['h_a'].mean()


security['v_m_up'] = security['v_m']+np.std(security['h_a'])*1.5
security['v_m_down'] = security['v_m']-np.std(security['h_a'])*1.5

security['tradeDate'] = pd.to_datetime(security['tradeDate'])
security = security.set_index('tradeDate')
security.info()

security['h_a'].plot(grid=False, figsize=(20,8))
security['v_m'].plot(grid=False, figsize=(20,8))
security['v_m_up'].plot(grid=False, figsize=(20,8))
security['v_m_down'].plot(grid=False, figsize=(20,8))


```
