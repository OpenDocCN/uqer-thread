# 随机选股

```python
#A股中随机抽取N只股票，返回List。
import random
import time
def stocks_random(N):
    universe=list(DataAPI.MktEqudGet(tradeDate=time.strftime("%Y%m%d"),field=u"secID",pandas="1").secID)
    stocks=set()
    while len(stocks)<N:
        stocks.add(universe[random.randint(0,len(universe)-1)])
    return list(stocks)
stocks_random(5)
```
