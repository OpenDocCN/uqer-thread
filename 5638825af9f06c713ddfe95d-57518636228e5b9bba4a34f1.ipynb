# 主题概念股票池探索

社区有帖子关注如何设置股票池为某个概念板块，笔者搜索了一下uqer的DataAPI使用帮助：

https://uqer.datayes.com/help/api/browse/datayes/master/?page=1

猛然发现，这不就有一个现成的主题概念API吗，** DataAPI.TickersByThemesGet **

## 1. 先看一下这个API做了什么事

```python
DataAPI.TickersByThemesGet(themeID=u"",themeName=u"二胎",beginDate=u"",endDate=u"",isNew=u"",field=u"",pandas="1")
```

#### 可以看出，这只API给出了相关的概念板块（'二胎'）包含的股票详情

上面表格中，最后一个（'刺激二胎'）是什么鬼呢？
- 此API对于主题概念的查询是模糊查找，所以不要担心会错过某些股票

## 2. 可是，themeName 里面可以填哪些概念呢？

我们来探索一下

```python
data = []

# 重复调用 DataAPI.TickersByThemesGet，得到所有当前可用的概念主题名称
for i in range(10):
    i_from = i * 1000
    i_to = (i + 1) * 1000
    try:
        tmp = DataAPI.TickersByThemesGet(themeID=map(str, range(i_from,i_to)),pandas="1")
        data = data + tmp.groupby('themeName').first().index.tolist()
    except:
        continue
```

```python
# 将主题概念名称打印出来
count = 0
for x in data:
    if count % 10 == 0:
        print ''
    count += 1
    
    print x, ' ',
```

#### Oh my Lady Gaga, 看起来好有内容啊！ 实际上，有上千个主题供你选用
