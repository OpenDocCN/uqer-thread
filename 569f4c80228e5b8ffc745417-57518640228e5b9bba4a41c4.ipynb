# 【大师系列】-11-罗伯特•山朋价值型投资法

### 罗伯特•山朋（ Robert Sanborn）是美国著名的价值型基金投资经理人。在投资组合选取中，罗伯特•山朋重视企业财务报表信息及公司治理能力。

> > `选股思路` ：


>     1.   健康的财务状况，财务报表数据分析 

>     2.   出色的公司治理水平，管理层持股比例

> > `选股条件` ：


>     1.   市销率为正且市销率低于市销率为正公司的平均值。  

>     2.   股价和每股企业自由现金流量比为正， 且低于所有该指标为正的公司的平均值。

>     3.   管理层持股比例高于平均值。


>    >  `参考资料` ：

>     《 申万宏源-申万大师系列.价值投资篇之十一：罗伯特•山朋价值型投资法 》
