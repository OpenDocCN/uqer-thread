```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
from matplotlib import pyplot as plt
import matplotlib

class amplitude_classification(object):    

    def __init__(self, secID='000001.XSHE', ticker='000001' , shift = 5, begin_date = u"20110101", end_date = u"20161030", 
                 frequency = 'half',dynamic = True, universe = ['000001.XSHE','000002.XSHE']):
        self.secID =secID
        self.ticker = ticker 
        self.universe = universe
        self.shift = shift
        self.begin_date = begin_date
        self.end_date = end_date
        self.frequency = frequency
        self.dynamic = dynamic
    
    def get_date_list(self):
        tmp = DataAPI.TradeCalGet(exchangeCD=u"XSHG", beginDate=self.begin_date, endDate=self.end_date, 
                                  field=['calendarDate','isOpen','isWeekEnd','isMonthEnd'])
        tmp = tmp[tmp['isOpen'] == 1]
        
        if self.frequency == 'daily':
            date = map(lambda x: x[0:4] + x[5:7] + x[8:10], tmp['calendarDate'].values.tolist())
        elif self.frequency == 'weekly':
            tmp = tmp[tmp['isWeekEnd'] == 1]
            date = map(lambda x: x[0:4] + x[5:7] + x[8:10], tmp['calendarDate'].values.tolist())
        elif self.frequency == 'monthly':
            tmp = tmp[tmp['isMonthEnd'] == 1]
            date = map(lambda x: x[0:4] + x[5:7] + x[8:10], tmp['calendarDate'].values.tolist())
        #half 有bug，index数字序列错误
        elif self.frequency == 'half':
            print '有bug，index数字序列错误，使用 get_date_list_shift()'
            tmp = tmp[tmp.index%10 == 0]
            date = map(lambda x: x[0:4] + x[5:7] + x[8:10], tmp['calendarDate'].values.tolist())
        return date
        
    def get_date_list_shift(self):
        tmp = DataAPI.TradeCalGet(exchangeCD=u"XSHG", beginDate=self.begin_date, endDate=self.end_date, 
                                  field=['calendarDate','isOpen'])
        tmp = tmp[tmp['isOpen'] == 1]
        tmp['No'] = range(len(tmp))
        tmp['sign'] = np.NAN
        del tmp['isOpen']
        tmp.set_index('No',inplace= True)
        shift = self.shift
        for i in tmp.index:
            if i % shift == 0:
                tmp.ix[i,'sign'] = 1
        tmp = tmp.dropna()
        date = map(lambda x: x[0:4] + x[5:7] + x[8:10], tmp['calendarDate'].values.tolist())        
        return date
    
    def get_multi_stock_return_one_day(self,date):
        date_list = self.get_date_list_shift()
        index = date_list.index(date)
        if index == 0:
            print 'wrong with index , the index cannot be 0 , the index now is ',index
        
        tmp=DataAPI.MktEqudAdjGet(tradeDate= date_list[index-1],secID= self.universe, field=u"secID,closePrice",pandas="1")
        tmp1=DataAPI.MktEqudAdjGet(tradeDate= date, secID= self.universe, field=u"secID,closePrice",pandas="1")
        tmp1.columns = ['secID','next_closePrice']
        tmp = tmp.merge(tmp1, on = 'secID')
        tmp['return'] = (tmp.next_closePrice/tmp.closePrice)-1.0
        tmp.set_index('secID',inplace=True)
        tmp= tmp['return'].dropna()
        return tmp
    
    def get_multi_stock_return_period(self):
        date_list = self.get_date_list_shift()
        tmp_collection = {}
        for i in date_list:
            if date_list.index(i) == 0:
                continue
            else:
                date = i
                index = date_list.index(i)
                tmp=DataAPI.MktEqudAdjGet(tradeDate= date_list[index-1],secID= self.universe, field=u"secID,closePrice",pandas="1")
                tmp1=DataAPI.MktEqudAdjGet(tradeDate= date, secID= self.universe, field=u"secID,closePrice",pandas="1")
                tmp1.columns = ['secID','next_closePrice']
                tmp = tmp.merge(tmp1, on = 'secID')
                tmp['return'] = (tmp.next_closePrice/tmp.closePrice)-1.0
                tmp.set_index('secID',inplace=True)
                tmp= tmp['return'].dropna()
                tmp_collection[i] = tmp
        return pd.DataFrame(tmp_collection)
    
    def get_multi_stock_return_period_count(self):
        m= {} 
        tmp = self.get_multi_stock_return_period()
        tmp = tmp.round(2)
        for i in tmp.index:
            tmp_i = pd.Series(tmp.ix[i].value_counts())
            m[i]= tmp_i 
        m = pd.DataFrame(m)
        return pd.Series(m.sum(axis=1))

    def single_index_return(self):
        tmp=DataAPI.MktIdxdGet(ticker= self.ticker, beginDate= self.begin_date, endDate= self.end_date, field=u"",pandas="1")
        tmp['x_shift_return'] = tmp.closeIndex.shift(-self.shift)/tmp.closeIndex -1.0
        tmp = tmp.dropna()
        tmp.set_index('tradeDate',inplace=True)
        tmp =  tmp['x_shift_return']
        return tmp
    
    def single_index_return_st(self):
        tmp=DataAPI.MktIdxdGet(ticker= self.ticker, beginDate= self.begin_date, endDate= self.end_date, field=u"",pandas="1")
        tmp['x_shift_return'] = tmp.closeIndex.shift(-self.shift)/tmp.closeIndex -1.0
        tmp = tmp.dropna()
        tmp['st_return'] = (tmp['x_shift_return'] - tmp['x_shift_return'].mean())/tmp['x_shift_return'].std()
        tmp.set_index('tradeDate',inplace=True)
        tmp = tmp['st_return'].dropna()
        return tmp
    
    def single_stock_return(self):
        tmp=DataAPI.MktEqudAdjGet(secID=self.secID ,isOpen="", beginDate= self.begin_date, endDate= self.end_date, field=u"",pandas="1")
        tmp = tmp[tmp.isOpen == 1]
        tmp['shift_return'] = (tmp.closePrice.shift(self.shift)/tmp.closePrice) - 1.0
        tmp.set_index('tradeDate',inplace=True)
        tmp = tmp['shift_return']
        tmp = tmp.dropna()
        return tmp
    
    def single_index_plot_distplot(self):
        tmp= self.single_index_return()
        fig = plt.figure(figsize=(16,7))
        sns.distplot(tmp,label= 'shift return',fit=stats.gamma)
  
    def single_index_plot_pie(self):
        tmp= self.single_index_return()
        ts = np.round(tmp,2).value_counts()
        fracs=ts.values
        labels=ts.index
        explode=np.zeros(len(labels))
        fig = plt.figure(1, figsize=(15,15)) #figsize=(6,6) 控制图的大小
        ax = plt.axes([0.1, 0.1, 0.8, 0.8])
        def autopct_func():
            return '%1.1f%%'
        res = ax.pie(fracs,
                     labels=labels,
                     explode=explode,
                     autopct=autopct_func(),
                     shadow=True, 
                     startangle=90)
        
    def single_stock_plot_distplot(self):
        tmp= self.single_stock_return()
        fig = plt.figure(figsize=(16,7))
        sns.distplot(tmp,label= 'shift return',fit=stats.gamma)
  
    def single_stock_plot_pie(self):
        tmp= self.single_stock_return()
        ts = np.round(tmp,2).value_counts()
        fracs=ts.values
        labels=ts.index
        explode=np.zeros(len(labels))
        fig = plt.figure(1, figsize=(15,15)) #figsize=(6,6) 控制图的大小
        ax = plt.axes([0.1, 0.1, 0.8, 0.8])
        def autopct_func():
            return '%1.1f%%'
        res = ax.pie(fracs,
                     labels=labels,
                     explode=explode,
                     autopct=autopct_func(),
                     shadow=True, 
                     startangle=90)
    
    def multi_stock_plot_pie(self):
        ts= self.get_multi_stock_return_period_count()
        fracs=ts.values
        labels=ts.index
        explode=np.zeros(len(labels))
        fig = plt.figure(1, figsize=(15,15)) #figsize=(6,6) 控制图的大小
        ax = plt.axes([0.1, 0.1, 0.8, 0.8])
        def autopct_func():
            return '%1.1f%%'
        res = ax.pie(fracs,
                     labels=labels,
                     explode=explode,
                     autopct=autopct_func(),
                     shadow=True, 
                     startangle=90)
        
    def mulit_stock_bar(self):
        tmp1=tmp.get_multi_stock_return_period_count()
        tmp2 =tmp1.sort_values()
        tmp3=tmp2.iloc[-20:]
        tmp4 = tmp3.sort_index()
        fig = plt.figure(figsize=(12,5))
        plt.bar(tmp4.index,tmp4.values,width=0.005)       
```

```python
import numpy as np
import pandas as pd
import scipy.stats as st

class multifactors_performance_on_basical_factor(object):
    def __init__(self, basical_factor, multifactors, begin_date, end_date,shift= 7, frequency = 'monthly', 
                 universe_choice = 'A', dynamic = False):
        self.begin_date = begin_date
        self.end_date = end_date
        self.frequency = frequency
        self.universe_choice = universe_choice
        self.basical_factor = basical_factor
        self.multifactors = multifactors
        self.dynamic = dynamic
        self.shift = shift
        #调用类函数生成时间列表
        self.date_list = self.get_date_list_shift()
        
        
    def get_date_list_shift(self):
        tmp = DataAPI.TradeCalGet(exchangeCD=u"XSHG", beginDate=self.begin_date, endDate=self.end_date, 
                                  field=['calendarDate','isOpen'])
        tmp = tmp[tmp['isOpen'] == 1]
        tmp['No'] = range(len(tmp))
        tmp['sign'] = np.NAN
        del tmp['isOpen']
        tmp.set_index('No',inplace= True)
        shift = self.shift
        for i in tmp.index:
            if i % shift == 0:
                tmp.ix[i,'sign'] = 1
        tmp = tmp.dropna()
        date = map(lambda x: x[0:4] + x[5:7] + x[8:10], tmp['calendarDate'].values.tolist())        
        return date
            
    def get_date_list(self):
        tmp = DataAPI.TradeCalGet(exchangeCD=u"XSHG", beginDate=self.begin_date, endDate=self.end_date, 
                                  field=['calendarDate','isOpen','isWeekEnd','isMonthEnd'])
        tmp = tmp[tmp['isOpen'] == 1]
        
        if self.frequency == 'daily':
            date = map(lambda x: x[0:4] + x[5:7] + x[8:10], tmp['calendarDate'].values.tolist())
        elif self.frequency == 'weekly':
            tmp = tmp[tmp['isWeekEnd'] == 1]
            date = map(lambda x: x[0:4] + x[5:7] + x[8:10], tmp['calendarDate'].values.tolist())
        elif self.frequency == 'monthly':
            tmp = tmp[tmp['isMonthEnd'] == 1]
            date = map(lambda x: x[0:4] + x[5:7] + x[8:10], tmp['calendarDate'].values.tolist())
        elif self.frequency == 'half':
            tmp = tmp[tmp.index%10 == 0]
            date = map(lambda x: x[0:4] + x[5:7] + x[8:10], tmp['calendarDate'].values.tolist())
        return date

    def get_divided_into_five_groups(self,break_point):
        basical_factor_field = 'secID,' + self.basical_factor
        if self.dynamic:
            tmp = DataAPI.MktStockFactorsOneDayGet(tradeDate = break_point, secID = set_universe(self.universe_choice, break_point), 
                                                  field = basical_factor_field).dropna()
        else:
            tmp = DataAPI.MktStockFactorsOneDayGet(tradeDate = break_point, secID = set_universe(self.universe_choice), 
                                                  field = basical_factor_field).dropna()
        tmp.sort(tmp.columns[1], ascending = True, inplace = True)
        
        secIDs = tmp['secID'].tolist()
        A05 = secIDs[:np.round(len(tmp)/20)]
        A10 = secIDs[:np.round(len(tmp)/10)]
        A1 = secIDs[:np.round(len(tmp)/5)]
        A2 = secIDs[np.round(len(tmp)/5): np.round(2*len(tmp)/5)]
        A3 = secIDs[np.round(2*len(tmp)/5): np.round(3*len(tmp)/5)]
        A4 = secIDs[np.round(3*len(tmp)/5): np.round(4*len(tmp)/5)]
        A5 = secIDs[np.round(4*len(tmp)/5):]
        A90 = secIDs[np.round(9*len(tmp)/10):]
        A95 = secIDs[np.round(19*len(tmp)/20):]
        
        return A05, A10, A1, A2, A3, A4, A5, A90, A95 
    
    def get_returns(self,secIDs,date):
        index = self.date_list.index(date)
        if index == 0:
            print 'wrong with index , the index cannot be 0 , the index now is ',index
        
        tmp_early = DataAPI.MktEqudAdjGet(tradeDate = self.date_list[index-1], secID = secIDs, field = u'secID,closePrice').dropna()
        tmp_current = DataAPI.MktEqudAdjGet(tradeDate = date, secID = secIDs, field = u'secID,closePrice').dropna()
        tmp = pd.merge(tmp_early, tmp_current, on= 'secID').dropna()
        tmp.set_index('secID',inplace = True)
        tmp.columns=['early_closePrice','current_closePrice']
        tmp['return'] = tmp['current_closePrice']/tmp['early_closePrice']- 1.0
        tmp = tmp['return']
        return tmp

    def get_returns_deal(self,secIDs,date):
        index = self.date_list.index(date)
        if index == 0:
            print 'wrong with index , the index cannot be 0 , the index now is ',index
        
        tmp_early = DataAPI.MktEqudAdjGet(tradeDate = self.date_list[index-1], secID = secIDs, field = u'secID,closePrice').dropna()
        tmp_current = DataAPI.MktEqudAdjGet(tradeDate = date, secID = secIDs, field = u'secID,closePrice').dropna()
        tmp = pd.merge(tmp_early, tmp_current, on= 'secID').dropna()
        tmp.set_index('secID',inplace = True)
        tmp.columns=['early_closePrice','current_closePrice']
        tmp['return'] = tmp['current_closePrice']/tmp['early_closePrice']- 1.0
        tmp['return']=pd.Series(standardize(neutralize(winsorize(tmp['return'].dropna().to_dict()),date)))  
        tmp = tmp['return']
        return tmp
    
    def get_multifactors(self,secIDs,date):
        #取多因子，并处理
        date = self.date_list[self.date_list.index(date)-1]
        multifactors_field = 'secID,' + self.multifactors
        tmp = DataAPI.MktStockFactorsOneDayGet(tradeDate = date, secID = secIDs, field = multifactors_field).dropna()
        tmp.set_index('secID', inplace = True)
        for i in tmp.columns:
            tmp[i]=pd.Series(standardize(neutralize(winsorize(tmp[i].dropna().to_dict()),date)))
        return tmp
    
    def get_rankIC(self):
        A05 = {}
        A10 = {}
        A1 = {}
        A2 = {}
        A3 = {}
        A4 = {}
        A5 = {}
        A90 = {}
        A95 = {}

        col_tmp = DataAPI.MktStockFactorsOneDayGet(tradeDate= '20160930', secID= '000001.XSHE', field = self.multifactors)
        for i in col_tmp.columns:
            A05[i] = []
            A10[i] = []
            A1[i] = []
            A2[i] = []
            A3[i] = []
            A4[i] = []
            A5[i] = []
            A90[i] = []
            A95[i] = []
            
        for i in range(1,len(self.date_list)):
            date = self.date_list[i]
            A05secIDs, A10secIDs, A1secIDs, A2secIDs, A3secIDs, A4secIDs, A5secIDs, A90secIDs,A95secIDs =self.get_divided_into_five_groups(date)
            
            returns_A05 = self.get_returns_deal(A05secIDs, date)
            returns_A10 = self.get_returns_deal(A10secIDs, date)
            returns_A1 = self.get_returns_deal(A1secIDs, date)
            returns_A1 = self.get_returns_deal(A1secIDs, date)
            returns_A2 = self.get_returns_deal(A2secIDs, date)
            returns_A3 = self.get_returns_deal(A3secIDs, date)
            returns_A4 = self.get_returns_deal(A4secIDs, date)
            returns_A5 = self.get_returns_deal(A5secIDs, date)
            returns_A90 = self.get_returns_deal(A90secIDs, date)
            returns_A95 = self.get_returns_deal(A95secIDs, date)
            
            multifactors_A05 = self.get_multifactors(A05secIDs, date)
            multifactors_A10 = self.get_multifactors(A10secIDs, date)
            multifactors_A1 = self.get_multifactors(A1secIDs, date)
            multifactors_A2 = self.get_multifactors(A2secIDs, date)
            multifactors_A3 = self.get_multifactors(A3secIDs, date)
            multifactors_A4 = self.get_multifactors(A4secIDs, date)
            multifactors_A5 = self.get_multifactors(A5secIDs, date)
            multifactors_A90 = self.get_multifactors(A90secIDs, date)
            multifactors_A95 = self.get_multifactors(A95secIDs, date)

            multifactors_A05['return'] = returns_A05 
            multifactors_A05 = multifactors_A05.dropna()
            multifactors_A10['return'] = returns_A10 
            multifactors_A10 = multifactors_A10.dropna()
            multifactors_A1['return'] = returns_A1 
            multifactors_A1 = multifactors_A1.dropna()
            multifactors_A2['return'] = returns_A2 
            multifactors_A2 = multifactors_A2.dropna()
            multifactors_A3['return'] = returns_A3 
            multifactors_A3 = multifactors_A3.dropna()
            multifactors_A4['return'] = returns_A4 
            multifactors_A4 = multifactors_A4.dropna()
            multifactors_A5['return'] = returns_A5 
            multifactors_A5 = multifactors_A5.dropna()
            multifactors_A90['return'] = returns_A90 
            multifactors_A90 = multifactors_A90.dropna()
            multifactors_A95['return'] = returns_A95 
            multifactors_A95 = multifactors_A95.dropna()
            
            for t in col_tmp.columns:
                tmpA05,_ = st.spearmanr(multifactors_A05['return'],multifactors_A05[t])
                tmpA10,_ = st.spearmanr(multifactors_A10['return'],multifactors_A10[t])
                tmpA1,_ = st.spearmanr(multifactors_A1['return'],multifactors_A1[t])
                tmpA2,_ = st.spearmanr(multifactors_A2['return'],multifactors_A2[t])
                tmpA3,_ = st.spearmanr(multifactors_A3['return'],multifactors_A3[t])
                tmpA4,_ = st.spearmanr(multifactors_A4['return'],multifactors_A4[t])
                tmpA5,_ = st.spearmanr(multifactors_A5['return'],multifactors_A5[t])
                tmpA90,_ = st.spearmanr(multifactors_A90['return'],multifactors_A90[t])
                tmpA95,_ = st.spearmanr(multifactors_A95['return'],multifactors_A95[t])
                
                A05[t].append(tmpA05)
                A10[t].append(tmpA10)
                A1[t].append(tmpA1)
                A2[t].append(tmpA2)
                A3[t].append(tmpA3)
                A4[t].append(tmpA4)
                A5[t].append(tmpA5)
                A90[t].append(tmpA90)
                A95[t].append(tmpA95)
        return A05, A10, A1, A2, A3, A4, A5, A90, A95, col_tmp.columns
    
    def analysis(self):
        A05, A10, A1, A2, A3, A4, A5, A90, A95, col_names = self.get_rankIC()
        result =pd.DataFrame(columns= col_names, index = np.arange(27),data=0)
        for i in col_names:
            mean_A05 = np.array(A05[i]).mean()
            mean_A10 = np.array(A10[i]).mean()
            mean_A1 = np.array(A1[i]).mean()
            mean_A2 = np.array(A2[i]).mean()
            mean_A3 = np.array(A3[i]).mean()
            mean_A4 = np.array(A4[i]).mean()
            mean_A5 = np.array(A5[i]).mean()
            mean_A90 = np.array(A90[i]).mean()
            mean_A95 = np.array(A95[i]).mean()
            
            std_A05 = np.array(A05[i]).std()
            std_A10 = np.array(A10[i]).std()
            std_A1 = np.array(A1[i]).std()
            std_A2 = np.array(A2[i]).std()
            std_A3 = np.array(A3[i]).std()
            std_A4 = np.array(A4[i]).std()
            std_A5 = np.array(A5[i]).std()
            std_A90 = np.array(A90[i]).std()
            std_A95 = np.array(A95[i]).std()
            
            result[i] = [mean_A05, mean_A10, mean_A1, mean_A2, mean_A3, mean_A4, mean_A5, mean_A90, mean_A95, std_A05, std_A10, std_A1, std_A2, std_A3, std_A4, std_A5, std_A90, std_A95, mean_A05/std_A05,  mean_A10/std_A10,  mean_A1/std_A1, mean_A2/std_A2, mean_A3/std_A3, mean_A4/std_A4, mean_A5/std_A5,  mean_A90/std_A90,  mean_A95/std_A95]
        result = result.T
        name_list = [['mean','mean','mean','mean','mean','mean','mean','mean','mean','std','std','std','std','std','std','std','std','std','IR','IR','IR','IR','IR','IR','IR','IR','IR'],['0~5%','0~10%','0~20%','20%~40%','40%~60%','60%~80%','80%~100%','90%~100%','95%~100%','0~5%','0~10%','0~20%','20%~40%','40%~60%','60%~80%','80%~100%','90%~100%','95%~100%','0~5%','0~10%','0~20%','20%~40%','40%~60%','60%~80%','80%~100%','90%~100%','95%~100%']]
        result.columns = pd.MultiIndex.from_tuples(zip(*name_list))
        return pd.concat([result.T],keys=[self.basical_factor+'分层后因子表现 时间范围：'+self.begin_date+'--'+self.end_date+'  数据频率：'+self.frequency+'  股票范围：' + self.universe_choice],axis=1)
    
    def signal_out(self):
        A05, A10, A1, A2, A3, A4, A5, A90, A95, col_names = self.get_rankIC()
        result =pd.DataFrame(columns= col_names, index = np.arange(27),data=0)
        for i in col_names:
            mean_A1 = np.array(A1[i]).mean()
            std_A1 = np.array(A1[i]).std()
        return mean_A1/std_A1
```

```python
import pandas as pd
import numpy as np
from datetime import datetime
from datetime import timedelta
from sklearn.svm import SVC
from CAL.PyCAL import *

def rebalance(account, buylist,ownlist):
    for stock in account.security_position:
        if stock not in buylist:
            if stock not in ownlist:
                try:
                    order_to(stock, 0)
                except:
                    print 'wrong with ',stock,' cannot sell'
    if len(buylist)<=10:
        length=10
    else:
        length=len(buylist)    
    
    for stock in buylist:
        try:
            order(stock, account.reference_portfolio_value / (1.1*length) / account.reference_price[stock])
        except:
            pass
    return   

def train_data(universe,used_factors,start = '2014-05-01',end = '2014-10-30',ad_period=10):
    cal = Calendar('China.SSE')
    date=cal.bizDatesList(start,end)
    set_factors=['tradeDate','secID'] 
    training_data_set=pd.DataFrame()
    for i,idate in enumerate(date):
        if i+ad_period>=len(date):
            continue
        t=i+ad_period
        tdate=date[t]
        tmp_price=DataAPI.MktEqudGet(tradeDate=idate,secID=universe,isOpen="",field=u"secID,tradeDate,closePrice",pandas="1")
        tmp_t_price=DataAPI.MktEqudGet(tradeDate=tdate,secID=universe,isOpen="",field=u"secID,tradeDate,closePrice",pandas="1")
        tmp_price['updown']=(tmp_t_price['closePrice']-tmp_price['closePrice'])/tmp_price['closePrice']
        # 设置涨跌幅度分级
        for i in tmp_price.index:
            if tmp_price.ix[i,['updown']].values>=0.08:
                tmp_price.ix[i,['updown']]=2
            elif tmp_price.ix[i,['updown']].values>=0.03 and tmp_price.ix[i,['updown']].values < 0.08:
                tmp_price.ix[i,['updown']]=1
            elif tmp_price.ix[i,['updown']].values<=0.03 and tmp_price.ix[i,['updown']].values>=-0.03:
                tmp_price.ix[i,['updown']]=0
            elif tmp_price.ix[i,['updown']].values<=-0.03 and tmp_price.ix[i,['updown']].values>=-0.08:
                tmp_price.ix[i,['updown']]=-1
            elif tmp_price.ix[i,['updown']].values<=-0.08:
                tmp_price.ix[i,['updown']]=-2
        tmp_price['updown']=tmp_price['updown'].astype(int)
        
        #因子参数
        tmp=DataAPI.MktStockFactorsOneDayGet(tradeDate=idate,secID=universe,field=set_factors+used_factors,pandas="1")
        #数据合并，因子处理
        tmp=pd.merge(tmp,tmp_price)
        tmp=tmp.set_index(['secID'])
        tmp=tmp.dropna()
        for ifac in used_factors:
            tmp[ifac]=pd.Series(standardize(neutralize(winsorize(tmp[ifac].to_dict(),n_draw=5),idate.strftime('%Y%m%d'))))  
        training_data_set=pd.concat([training_data_set,tmp])
    #重新设置index
    rownum=range(len(training_data_set))
    rownum=pd.Series(rownum)
    training_data_set=training_data_set.set_index(rownum)  
    training_data_set=training_data_set.dropna()

    Xtrain=training_data_set.loc[:,used_factors]
    Xtrain=np.mat(Xtrain)
    Ytrain=training_data_set.loc[:,'updown']
    Ytrain=np.mat(Ytrain).T
    return Xtrain,Ytrain    
```
