```python
import pandas as pd
import datetime
import time
tradeDate = '2021-08-13'
data=DataAPI.MktStockFactorsOneDayGet(tradeDate=tradeDate ,ticker=u"",\
                                       field=u"ticker,pb,pe,MA5,VOL5",pandas="1")

data1=DataAPI.EquGet(ticker=u"",equTypeCD=u"A",listStatusCD=u"",
                     field=u"ticker,secShortName,totalShares",pandas="1")
data1['ticker_head3']=data1.apply(lambda x:str(x.ticker)[0:3], axis = 1)
#去掉科创板
data1=data1[data1['ticker_head3']!="688"]
#去掉ST
data1=data1[~data1['secShortName'].str.contains("ST")]
totalShares=3
VOL5=0.02
MA5=30
data=data[(data.VOL5>VOL5)&(data.MA5<MA5)]
data1=data1[data1.totalShares<=totalShares*100000000]
res=pd.merge(data,data1,on="ticker")
#res=res.drop_duplicates('ticker',keep=False) #删除ticker列重复值，返回副本 错！！！把重复的删除了
res=res.drop_duplicates('ticker',keep='first',inplace=False)#20210506 发现无002566这才对
print "%s:去掉科创板及ST,总股本小于%d亿,VOL5大于%.2f及MA5小于%d:" %(tradeDate,totalShares,VOL5,MA5)
#print res
tickers=res["ticker"].values
print tickers
```

```python
import pandas as pd
import scipy.stats as stats
import datetime
import time
from CAL.PyCAL import *
cal= Calendar('China.IB')
ticker_bmark="002171"
beginDate="20210616"
endDate="20210716"
price_bmark=DataAPI.MktEqudAdjGet(ticker=ticker_bmark,beginDate=beginDate,endDate=endDate,isOpen="1",\
                            field=u"tradeDate,closePrice,turnoverValue,turnoverVol",pandas="1")
price_bmark['meanPrice']=price_bmark.apply(lambda x:x.turnoverValue/x.turnoverVol, axis = 1)
price_bmark=price_bmark[['meanPrice','tradeDate']]
#price_bmark=price_bmark[['closePrice','tradeDate']]
price_bmark.plot()
print price_bmark  
price_bmark=[stk[0] for stk in price_bmark.values]
length=len(price_bmark)
print price_bmark
today="20210801"
today = datetime.datetime.strptime(today,"%Y%m%d") #将日期化为API可以识别的样子，也可用CAL
today=today.strftime("%Y%m%d")
#today = datetime.date.today().strftime("%Y%m%d") # 获得今天的日期
print "today:",today
#+20为了保证beginDate足够
totallength=length+20
beforeday="-"+str(totallength)+"B"
    #为了缩短取数据时间，精确获得开始日期
    # -80B 表示向前80个交易日，同理，5B表示向后5个交易日
beginDate = cal.advanceDate(today , beforeday, BizDayConvention.Following).strftime("%Y%m%d")
data=DataAPI.MktEqudAdjGet(ticker=tickers,beginDate=beginDate,endDate=today,isOpen="1",
        field=u"ticker,tradeDate,turnoverValue,turnoverVol",pandas="1")
data['meanPrice']=data.apply(lambda x:x.turnoverValue/x.turnoverVol, axis = 1)
data=data[['meanPrice','tradeDate','ticker']]
#print data

df = pd.DataFrame(columns=["ticker","r","p"])
num=0
for stock in tickers:
    datatmp=data[data.ticker==stock]
    datatmp=datatmp.tail(length)
    #print datatmp
    pricetmp=[stk[0] for stk in datatmp.values]
    #print pricetmp 
    r="NULL"
    p="NULL"
    if len(price_bmark)==len(pricetmp):
        [r,p]=stats.pearsonr(price_bmark,pricetmp)
        #print r,p
        df.loc[num]=[stock,r,p]
        num=num+1
df=df[df.p<0.05]
df=df.sort(["r"],ascending=False)
print "2021-08-13:去掉科创板及ST,总股本小于3亿,VOL5大于0.02及MA5小于30"
print "与股票：%s,从%s到%s价格相似，%s：的股票为:" %(ticker_bmark,beginDate,endDate,today)
print df.head(5)

```
