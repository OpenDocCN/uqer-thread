## 策略说明
1. 股票池为沪深300
2. 选取股息率最高的10家公司作为投资组合
3. 回测日期：2015-05-04至2019-06-30
4. 调仓日为5月第一个交易日

```python
import pandas as pd
import numpy as np
import datetime
from dateutil.parser import parse
```

## 1 编写时间截面函数
获得指定时间股息率前十的股票代码

```python
def chooseStock(universe, date):
    trade_date = date if isinstance(date, datetime.datetime) else parse(date)
    trade_date = trade_date.strftime('%Y%m%d')
    # 获取股息率
    df_factor = DataAPI.MktStockFactorsOneDayGet(secID=universe, tradeDate=trade_date, field=['secID', 'CTOP'])
    df_factor.sort_values('CTOP', ascending=False, inplace=True) # CTOP:现金流市值比
    # 选取前10只
    df_factor_select = df_factor.head(10)
    sec_list = df_factor_select['secID'].tolist()
    return sec_list
```

## 2 获取调仓日

```python
tradedaylist = DataAPI.TradeCalGet(exchangeCD=['XSHE', 'XSHG'])
tradedaylist = tradedaylist[tradedaylist['isOpen']==1]
tradedaylist = tradedaylist[tradedaylist['calendarDate']>'2007-01-01']
tradedaylist['month'] = tradedaylist['calendarDate'].apply(lambda x: x[5:7])
tradedaylist['year'] = tradedaylist['calendarDate'].apply(lambda x: x[:4])
tradedaylist =tradedaylist.drop_duplicates(subset=['month', 'year'], keep='first')
t_date = tradedaylist.ix[tradedaylist['month'].isin(['05']), :]['calendarDate'].values
t_date = [parse(x) for x in t_date]
t_date
```
