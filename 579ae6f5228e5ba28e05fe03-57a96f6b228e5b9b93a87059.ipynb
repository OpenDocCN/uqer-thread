# `概念`
***

- 相关性
	- 	`相关性(Correlation)`在统计中是与`独立性(independence)`对立的概念。泛指两随机变量之间存在的一切关联，但也经常被狭义地指代线性相关性。统计上只要不是相互独立的随机变量都被成为相关的；在实践中，可以通过计算相关系数的办法来量化相关性的大小。下文将介绍`Pearson相关系数`和`Spearman相关系数`的计算方法和特点。
  
  
- Pearson相关系数 
	- `Pearson相关系数`是对两随机变量之间**线性相关**关系的一种计量，在[-1, 1]区间上取值，1代表完全正（线性）相关，-1代表完全负（线性）相关，0代表线性无关。值得注意的是Pearson相关系数为0并不一定意味着随机变量独立，这一推断只有在随机变量服从联合正态分布时才成立。
	- 计算公式：
 \[r={\frac {\sum _{i=1}^{n}(x_{i}-{\bar {x}})(y_{i}-{\bar {y}})}{{\sqrt {\sum _{i=1}^{n}(x_{i}-{\bar {x}})^{2}}}{\sqrt {\sum _{i=1}^{n}(y_{i}-{\bar {y}})^{2}}}}}\]

- Spearman秩相关系数
    - `Spearman秩相关系数`是秩相关系数的一种，采用非参数的方法衡量两随机变量之间的单调相关关系（不一定是线性），也在[-1, 1]上取值，取1表示一随机变量随另一随机变量严格递增，-1表示严格递减。
    - 计算公式：
 \[r_{s}={1-{\frac {6\sum d_{i}^{2}}{n(n^{2}-1)}}}\]
 其中$ d_{i}=\operatorname {rg} (X_{i})-\operatorname {rg} (Y_{i})$，表示一个观测$(X_i,Y_i)$中两者之间秩的差距。
  
  
- 主成分分析法
    - `主成分分析法(PCA)`是降维分析法中较为简单的一种，是指对于多维数据，提取线性组合使得该线性组合能解释大部分原先数据的变化（方差）。
  
  
# `操作流程`
***
  
- `分组：`如同优矿因子日报和周报中的内容，现将因子按照风格或经济学含义不同分为**收入因子、规模因子、技术因子**等组合；
- `筛选：`计算各组合内部因子的相关系数，在高度相关的因子中挑选代表因子留下，其余剔除，保证各组合中剩余因子相关性不高；
- `降维(PCA)：`采用主成分分析法，计算能够代表各因子组合的第一主成分；
- `预测：`利用上一步计算所得的各个因子第一主成分预测股票价格走势。

```python
# built-in
from __future__ import division
import datetime
import time 
import pandas as pd
import numpy as np
import math
import string

# 3rd-party
import matplotlib
import matplotlib.pyplot as plt
from matplotlib import dates
matplotlib.style.use('ggplot')
import seaborn as sns
sns.set_style('white')
from matplotlib import rc
rc('mathtext', default='regular')
from CAL.PyCAL import *
from sklearn import linear_model
from sklearn.decomposition import PCA

# warm-up 
DataAPI.settings.cache_enabled = True
_TICKER = '000002'
```

```python
def handle_NaN_in_df(df, tolerance_ratio = 0.2):
    cols = []
    for x in df.columns:
        if len([1 for _ in df[x] if np.isnan(_)])/len(df) > tolerance_ratio:
            cols += [x]
    df = df[[_ for _ in df.columns if not _ in cols]]
    return df.fillna(0)


def factor_corr(factors, ticker = _TICKER, beginDate = '20120101', endDate = '', method = 'pearson'):
#     获取某只股票某段时间内一个因子组合的相关性矩阵，并作图；
    df = DataAPI.MktStockFactorsDateRangeGet(ticker = ticker, beginDate = beginDate,
                                             endDate= endDate, field = ['tradeDate'] + factors).set_index('tradeDate')
    res = df.corr(method = method)
    _ = sns.corrplot(df)
    return res


def factor_filter(corr_df, threshold = 0.6):
    f_filter = []
    for col,row in [(s, t) for s in range(len(corr_df)) for t in range(s+1,len(corr_df))]:
        if (col in f_filter) or (row in f_filter):
            continue
        if abs(corr_df[corr_df.index[col]][corr_df.index[row]]) > threshold:
            f_filter += [row]
    return [corr_df.index[s] for s in range(len(corr_df)) if not s in f_filter]


def sec_factor_filter(factors, ticker = _TICKER, beginDate = '20120101', endDate = ''):
#     给定某只股票与因子组合，筛选出相关性不高的代表因子组合；
    df = DataAPI.MktStockFactorsDateRangeGet(ticker = ticker, beginDate = beginDate, endDate = endDate, 
                                             field = ['tradeDate'] + factors).set_index('tradeDate')
    df = handle_NaN_in_df(df)
    return factor_filter(df.corr(method = 'pearson'))


def demension_deduction(factors,ticker = _TICKER, beginDate = '20120101', endDate = ''):
#     给定某股票与因子组合，提供能够解释该因子组合大部分变化（方差）的主成分数据。
    df = DataAPI.MktStockFactorsDateRangeGet(ticker = ticker, beginDate = beginDate, endDate = endDate, 
                                             field = ['tradeDate'] + factors).set_index('tradeDate')
    df = handle_NaN_in_df(df)
    pca_data = df[factor_filter(df.corr(method = 'pearson'))]
    pca = PCA(n_components= 1)
    pca.fit(pca_data)
    return pd.DataFrame(pca.fit(pca_data).transform(pca_data), columns = ['first_comp'], index = pca_data.index)

def df_standardize(df):
    for x in df.columns:
        df[x] = df[x] - df[x].mean()
        df[x] = df[x]/df[x].std()
    return df
```

```python
revenue_factors = [ 'NetProfitRatio', 'OperatingProfitRatio', 'GrossIncomeRatio', 
                   'ROA', 'ROA5', 'ROE', 'ROE5', 'EGRO', 'EPS', 'DilutedEPS' ]
scale_factors = [ 'CurrentRatio', 'QuickRatio', 'DebtEquityRatio', 'FixAssetRatio', 'IntangibleAssetRatio' ]
grow_factors = [ 'NetAssetGrowRate', 'TotalAssetGrowRate', 'OperatingRevenueGrowRate',
                 'OperatingProfitGrowRate', 'NetProfitGrowRate', 'NPParentCompanyGrowRate' ]
evaluation_factors = [ 'PE', 'PS', 'PCF', 'PB', 'ASSI' ]
cash_factors = [ 'CashRateOfSales', 'CashToCurrentLiability', 'FinancingCashGrowRate',
                                                'OperCashGrowRate', 'InvestCashGrowRate' ]
operation_factors = [ 'InventoryTRate', 'ARTRate', 'AccountsPayablesTRate', 
                       'CurrentAssetsTRate', 'TotalAssetsTRate' ]
analyst_factors = [ 'REC', 'FY12P', 'SFY12P' ]
tech_factors = [ 'MA5', 'MA10','MA20','MA60','MA120', 'VOL5',
                'VOL10', 'VOL20', 'VOL60', 'VOL120', 'EMA5', 'EMA10', 'EMA20', 'EMA60', 'EMA120',
                'REVS5', 'REVS10', 'REVS20','DAVOL5', 'DAVOL10', 'DAVOL20' ]

styles = ['revenue', 'scale', 'grow', 'evaluation', 'cash', 'operation', 'analyst', 'tech']
factor_dict = {'revenue': revenue_factors, 'scale': scale_factors, 'grow': grow_factors, 'evaluation': evaluation_factors,
    'cash': cash_factors, 'operation': operation_factors, 'analyst': analyst_factors, 'tech':tech_factors}
```

- 以`收入因子(revenue)`组合为例，画出一段时期以来各因子的相关系数矩阵。可以看见，一些因子的相关性非常高，把这些因子都加入统计模型可能浪费计算资源，如果加入线性回归还可能造成多重共线性的问题，于是考虑用sec_factor_filter选取组合代表因子。

```python
_ = plt.figure(figsize = (10, 10))
_ = factor_corr(factor_dict['revenue'])
```

- `sec_factor_filter`从收入因子组合中过滤出['NetProfitRatio', 'ROA', 'EPS']三个因子。

```python
sec_factor_filter(factor_dict['revenue'])
```

- 用主成分分析法(PCA)对各组合内代表因子数据进行降维并进行标准化，得到组合信号数据：

```python
first_comp_df = pd.DataFrame()
for sty in factor_dict:
    first_comp_df[sty] = df_standardize(demension_deduction(factor_dict[sty])).first_comp
first_comp_df.head(2)
```

- 利用因子组合信号数据对股价进行拟合，这里采用的是线性回归模型和OLS方法：

```python
ht = DataAPI.MktEqudGet(ticker = _TICKER, beginDate = '20120101', field = 'tradeDate,closePrice,preClosePrice').set_index('tradeDate')
ht['daily_return'] = ht.closePrice / ht.preClosePrice - 1
reg_data = pd.DataFrame.merge(first_comp_df, ht, left_index = True, right_index= True)
# reg_data.drop(['closePrice','preClosePrice'], inplace = True, axis = 1)


# Create linear regression object
regr = linear_model.LinearRegression()
regr.fit(reg_data[factor_dict.keys()], reg_data.closePrice) 
# sns.corrplot(reg_data)
```

- 画出预测/实际图像，考察预测效果。可见，在几次股价大幅变动中，所预测的股价都有提前的同方向变动，可以作为不错的信号，考虑进$alpha$策略中。
- 所不足的地方在于财务因子数据不是每天更新，需要等到财报公布时期，所以对于股价绝对值的预测可能并不非常准确，只能作为趋势判断信号。

```python
print "Regression R-square is %f"%regr.score(reg_data[factor_dict.keys()], reg_data.closePrice)
reg_data['prediction'] = regr.predict(reg_data[factor_dict.keys()])
_ = reg_data[['prediction','closePrice']].plot(figsize = (12, 12*0.618))
```

# `总结`
***
  
- 怎么利用因子数据预测行情是广大$alpha$追寻者关心的问题，本篇提供了我的一个思路，供大家参考。总结起来就是采用不同因子组合、剔除高度相关因子后再进行降维处理，利用所得各风格信号预测股价走势。
- 本篇以万科A(000002)和[因子周报](https://uqer.datayes.com/community/share/575e09b1228e5b8a3d28698a)中的因子组合作为示例，展示了分析流程与结果。在多次股价的大变动中，预测股价展现了较好的预示作用，足够充当一个判断行情走向的信号。
- 文中提供的接口是：
    - factor_corr：获取某只股票某段时间内一个因子组合的相关性矩阵，并作图；
    - sec_factor_filter：给定某只股票与因子组合，筛选出相关性不高的代表因子组合；
    - demension_deduction：给定某股票与因子组合，提供能够解释该因子组合大部分变化（方差）的主成分数据。
