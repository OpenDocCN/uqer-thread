解决方案：按照如下步骤进行操作：
这里所有财报的数据都是年初至财报期的累积数据
a.确定计算的是截止到哪天为止的TTM净利润值，记为T日
b.只取T日及之前发布的的财报
c.选取最近一个财报期的财报，记录其财报期为A
d.选取A，A之前一年的年报记为B，去年与A同期的财报记为C
e.截至T日TTM净利润 = A中净利润 + B中净利润 – C中净利润

````python
计算贵州茅台与恒瑞医药2018年2月所有交易日的TTM营业总收入和TTM管理费用
````

````python
import numpy as np
import pandas as pd

#确定计算的是截止到哪天为止的TTM净利润值，记为T日
T = pd.to_datetime('2018-02-28')  #截止日期


def get_df(ticker_symbol):
    #获取财报数据
    columns = [u'ticker', u'publishDate', u'endDate', u'fiscalPeriod', u'tRevenue', u'adminExp']
    MT = DataAPI.FdmtISGet(ticker = ticker_symbol, field = columns, pandas = "1")

    #统一时间格式
    MT['publishDate'] = pd.to_datetime(MT['publishDate'])
    MT['endDate'] = pd.to_datetime(MT['endDate'])
    MT = MT.loc[MT['endDate'].dt.month == MT['fiscalPeriod'].astype(int)]

    #只取T日及之前发布的的财报，保证不会偷窥未来
    df_MT = MT.loc[MT['publishDate'] <= T]
    return df_MT

def get_statement_date(ticker_symbol):
    df = get_df(ticker_symbol)
    df = df.sort_values(['endDate', 'publishDate'], ascending=[False, False])#按照财报期和发布期排序
    latest_date = df.iloc[0].loc['endDate']
    #获取去年及去年同期财报的endDate
    last_year_end_date = pd.to_datetime(str(latest_date.year - 1) + '1231')
    last_same_season_end_date = pd.to_datetime(int(latest_date.strftime('%Y%m%d')) - 10000, format='%Y%m%d')
    return last_year_end_date, last_same_season_end_date
# #计算TTM值

def get_ttm(ticker_symbol, T, col_name):
    df = get_df(ticker_symbol)
    #根据历史财报数据df，截止日期T，财务指标名（即净利润还是营收等），构造截至T日的TTM财务指标
    sec_num = df['ticker'].iloc[0]
    df = df.loc[df[u'publishDate'] <= T].copy()
    df = df.sort_values([u'endDate', u'publishDate'], ascending=[False, False])
    latest_date = df.iloc[0].loc[u'endDate']
    last_year_end_date, last_same_season_end_date = get_statement_date(ticker_symbol)
    value_this_season = df.loc[df[u'endDate'] == latest_date]
    if len(df) > 0:
        value_this_season = value_this_season.iloc[0].loc[col_name]
    else:
        value_this_season = np.nan
    value_last_year = df.loc[df[u'endDate'] == last_year_end_date]
    if len(value_last_year) > 0:
        value_last_year = value_last_year.iloc[0].loc[col_name]
    else:
        value_last_year = np.nan
    value_last_same_season = df.loc[df[u'endDate'] == last_same_season_end_date]
    if len(value_last_same_season) > 0:
        value_last_same_season = value_last_same_season.iloc[0].loc[col_name]
    else:
        value_last_same_season = np.nan
    value_TTM = value_this_season + value_last_year - value_last_same_season    
    return value_TTM

MT_tRVENUE = get_ttm(ticker_symbol, T, [u'tRevenue', u'adminExp'])

#获取交易日期
columns = ['calendarDate', 'isOpen']
trade_cal = DataAPI.TradeCalGet(exchangeCD = u"XSHG", beginDate=u"20180201",endDate=u"20180228", field = columns, pandas="1")
trade_cal = trade_cal[trade_cal['isOpen']==1]
trade_cal['calendarDate'] = pd.to_datetime(trade_cal['calendarDate'])
trade_cal = trade_cal.loc[trade_cal['calendarDate'].between(pd.to_datetime('20180201'), pd.to_datetime('20180228'))]

#生成结果
f_list = []
tickera_list = ['600519', '600276']
for ticker_symbol in tickera_list:
    for date in trade_cal['calendarDate']:
        MT_tRVENUE = get_ttm(ticker_symbol, T, [u'tRevenue', u'adminExp'])
        f_list.append([ticker_symbol, date, MT_tRVENUE[0], MT_tRVENUE[1]])
f_data  = pd.DataFrame(f_list, columns = ['ticker', 'tradeDate', 'tRevenue', 'adminExp'])

f_data
````

计算2018年2月份所有交易日的贵州茅台与恒瑞医药的Administrative expense rate因子值，并与优矿的比较。

备注：Administrative expense rate（管理费用与营业总收入之比）= 管理费用（TTM）/ 营业总收入（TTM）。

````python
#正确的取法
ticker_symbol = ['600519', '600276']
columns = ['ticker', 'publishDate', 'endDate', 'tRevenue', 'adminExp']
ttm_data_uqer = DataAPI.FdmtISInduTTMPITGet(ticker = ticker_symbol, endDate=u"", publishDateBegin = u"20170101", publishDateEnd = u"20180228", field = columns, pandas = "1")
ttm_data_uqer = ttm_data_uqer.sort_values(['ticker', 'publishDate', 'endDate'])
ttm_data_uqer = ttm_data_uqer.drop_duplicates(['ticker', 'publishDate'], keep='last')
ttm_data_uqer = ttm_data_uqer.rename(columns={'publishDate':'tradeDate'})
ttm_data_uqer = ttm_data_uqer[['ticker', 'tradeDate', 'tRevenue', 'adminExp']]
ttm_data_uqer['tradeDate'] = pd.to_datetime(ttm_data_uqer['tradeDate'])

# #获取交易日历
trade_cal = DataAPI.TradeCalGet(exchangeCD=u"XSHG",beginDate=u"",endDate=u"",field=u"isOpen,calendarDate",pandas="1")
trade_cal = trade_cal[trade_cal['isOpen']==1]
trade_cal['calendarDate'] = pd.to_datetime(trade_cal['calendarDate'])
trade_cal = trade_cal.loc[trade_cal['calendarDate'].between(pd.to_datetime('20170201'), pd.to_datetime('20180228'))]
trade_cal = trade_cal.rename(columns={'calendarDate':'tradeDate'}) 

#补充交易日期
ttm_data_uqer = ttm_data_uqer.set_index(['tradeDate', 'ticker']).unstack()
for date in trade_cal['tradeDate']:
    if date not in ttm_data_uqer.index:
        ttm_data_uqer.loc[date, :] = np.nan

#fillna，计算因子
ttm_data_uqer = ttm_data_uqer.sort_index().fillna(method='ffill')
ttm_data_uqer = ttm_data_uqer.stack().reset_index()
ttm_data_uqer['AdminiExpenseRate'] = ttm_data_uqer['adminExp'] / ttm_data_uqer['tRevenue']


ttm_data_uqer
````
