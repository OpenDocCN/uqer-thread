# 一、MACD 的DEA 指标价格分段简介
____________

发现价格走势规律的一个重要途径是基于特定规则对价格走势进行分段。任何投资交易都是基于低买高卖，如果可以做空，则高点做空低点买回，所以投资交易成功的关键在于把握价格走势的拐点，把握的波段级别越小，交易的回报率就越高，但对资金的容纳量也就越低。对已有价格走势进行分段之后，通过分析各个时间级别的价格拐点特征，可以提高预判当下拐点出现的择时能力。

价格分段的方法有很多，本文主要在uqer上实现基于MACD的DEA指标波段划分。

# 二、MACD的DEA指标波段划分
-----------------------------------------
MACD（Moving Average Convergence Divergence）称为指数平滑移动平均线，是技术分析领域应用广泛的指标之一，包含了三个参数，常用的设置有（12，26，9）。MACD是计算两条不同速度（长期与中期）的指数平滑移动平均线（EMA）的差离状况来作为研判行情的基础。DIF为12周期均值与26周期均值之差，DEA为DIF的9周期均值，而MACD则为DIF与DEA差值的两倍。

DEA 指标在价格上涨和下跌的过程中其穿越零轴这个特征对价格进行分段，没有使用DIF的原因是DIF走势稳定性较DEA差，价格上一些小的扰动会引起DIF在零轴上下穿越，而在这一点上DEA 稳定得多。

基于MACD 的DEA 指标波段划分规则：

① 波段低点对应MACD 的DEA 指标 < 0，波段顶点对应MACD 的DEA 指标 > 0；

② 每个波段中价格的最大值和最小值只出现在波段的起点或终点。

# 三、基于MACD的DEA指标分段拐点识别算法
--------
可疑底部拐点满足四个条件：

（1） DEA < 0；

（2） 向右遍历，直到找到满足两个条件的极大值点RT（① DEA>0；
② 这个极大值点是底点到此极大值点区间的最大值）；

（3） 向左遍历，直到找到满足两个条件的极大值点LT（① DEA>0；
② 这个极大值点是底点到此极大值点区间的最大值）；

（4） 那么位于LT 与RT 之间的最小值点就是可疑底部拐点。

可疑顶部拐点满足四个条件：

（1） DEA > 0；

（2） 向右遍历，直到找到满足两个条件的极小值点RB（① DEA < 0；
② 这个极小值点是顶点到此极小值点区间的最小值）；

（3） 向左遍历，直到找到满足两个条件的极小值点LB（① DEA < 0；
② 这个极小值点是顶点到此极小值点区间的最小值）；

（4） 那么位于LB 与RB 之间的最大值点就是可疑顶部拐点。

# 四、异常波段处理
______
刚才的找拐点的算法，并不能完全保证一个波段中的最大值一定是可疑顶部拐点且最小值一定是可疑底部拐点。所以当出现波段的最大值不是可疑顶部拐点或者最小值不是可疑底部拐点时，这样的波段就不是符合DEA 指标分段规则的波段。

具体异常的处理分类过多，在此不一一讨论，详见[研报](http://bbs.pinggu.org/thread-3594952-1-1.html)

特别的说一下，当DEA比较耦合时，还会出现研报没考虑的意外情况。譬如找到连续的底部拐点或找到连续的顶部拐点，这时候直接去掉中间的可疑拐点就行了。

```python
import talib as ta
class pieceWise(object):
    def __init__(self,series):
            
        self.DEA = ta.MACD(series)[1]  #选用默认设置（12，26，9）
        self.series = list(series)
        self.length = len(self.series)
        self.point = range(1,len(self.series))
        self.vertex = []  #顶拐点
        self.nadir = []   #底拐点
        self.inflexion = []  #拐点
        self.left = None
        self.right = None
    
    def isMinima(self,point1,point2):   # 判断某个备选顶点左右是否存在DEA小于零的极小值点
        
        if point1 < point2:
            if self.series[point1-1] > self.series[point1] and self.series[point1+1] > self.series[point1] and min(self.series[point1:point2+1]) ==self.series[point1] and self.DEA[point1] < 0:
                #print point1,self.DEA[point1] 
                return True
            else:
                return False
        else:
            if self.series[point1-1] > self.series[point1] and self.series[point1+1] > self.series[point1] and min(self.series[point2:point1+1]) ==self.series[point1] and self.DEA[point1] < 0:
                return True
            else:
                return False
            
    def isMaxima(self,point1,point2):  # 判断某个备选底点左右是否存在DEA大于零的极大值点
        
        if point1 < point2:
            if self.series[point1-1]<self.series[point1] and self.series[point1+1]<self.series[point1] and max(self.series[point1:point2+1]) ==self.series[point1] and self.DEA[point1] > 0:
                return True
            else:
                return False
        else:
            if self.series[point1-1] < self.series[point1] and self.series[point1+1] < self.series[point1] and max(self.series[point2:point1+1]) ==self.series[point1] and self.DEA[point1] > 0: 
                return True
            else:
                return False
            
    def isNadir(self,point):     # 判断该点是否为可疑底点
        
        flag_left = False
        flag_right = False
        
        for i in range(point-1,-1,-1):
            if self.isMaxima(i,point):
                self.left = i
                flag_left = True
                break
                
        for j in range(point+1,self.length-1):
            if self.isMaxima(j,point):
                self.right = j
                flag_right = True
                break
        
        if flag_left and flag_right:
            return True
        else:
            return False
    
    def isVertex(self,point):   #判断该点是否为可以顶点
        
        flag_left = False
        flag_right = False
        
        for i in range(point-1,-1,-1):
            if self.isMinima(i,point):
                #print 'a'
                self.left = i
                flag_left = True
                break
                
        for j in range(point+1,self.length-1):
            #print j
            if self.isMinima(j,point):
                #print 'b'
                self.right = j
                flag_right = True
                break
        
        if flag_left and flag_right:
            return True
        else:
            return False
        
    def getInflexion(self):              ## 得到可疑拐点
        
        for i in self.point:
                if self.DEA[i] > 0 and self.isVertex(i):

                    j = self.series.index(max(self.series[self.left:self.right]))
                    #print 'v',i,self.left,self.right,j
                    if j==i:
                        self.vertex.append(j)  ## 得到可疑顶点
                elif self.DEA[i] < 0 and self.isNadir(i):

                    k = self.series.index(min(self.series[self.left:self.right]))
                    #print 'n',i,self.left,self.right,k
                    if k==i:
                        self.nadir.append(k)  ## 得到可疑底点
                else:
                    pass

    def handleFallMax(self,interval,inflexion):   ###处理下跌波段最大值不是可疑顶点
        
        if inflexion.index(interval[0])==0:
            self.inflexion.remove(interval[0])

        elif inflexion.index(interval[1]) ==len(inflexion) -1:
            delte_point = max(self.series[inflexion[inflexion.index(interval[0]) - 1]],self.series[interval[1]])
            self.inflexion.remove(self.series.index(delte_point))

        else:
            if self.series[interval[0]] > self.series[inflexion[inflexion.index(interval[1]) + 1]]:
                self.inflexion.remove(inflexion[inflexion.index(interval[0]) - 1])
                self.inflexion.remove(interval[0])
            
            elif self.series[interval[0]] < self.series[inflexion[inflexion.index(interval[1]) + 1]] and self.series[inflexion[inflexion.index(interval[0]) - 1]] < self.series[interval[1]]:
                self.inflexion.remove(interval[0])
                self.inflexion.remove(interval[1])
                
            elif self.series[interval[0]] < self.series[inflexion[inflexion.index(interval[1]) + 1]] and self.series[inflexion[inflexion.index(interval[0]) - 1]] > self.series[interval[1]]:
                self.inflexion.remove(inflexion[inflexion.index(interval[0]) - 1])
                self.inflexion.rmove(interval[0])
                
    def handleUpMin(self,interval,inflexion):          ###处理上涨波段最小值不是可疑底点
        self.series = [1.0/item for item in self.series]
        self.handleFallMax(interval,inflexion)
        self.series = [1.0/item for item in self.series]     
            
    def handleFallMin(self,interval,inflexion):        ###处理下跌波段最小值不是可疑底点
        
        if inflexion.index(interval[0])==0:
            delte_point = min(self.series[inflexion[inflexion.index(interval[1]) + 1]],self.series[interval[0]])
            self.inflexion.remove(self.series.index(delte_point))
        elif self.inflexion.index(interval[1]) == (len(self.inflexion)-1):
            self.inflexion.remove(interval[1])

        else:
            if self.series[interval[0]] < self.series[inflexion[inflexion.index(interval[1]) + 1]] and self.series[inflexion[inflexion.index(interval[0]) - 1]] < self.series[interval[1]]:
                #print 'dot_one'
                self.inflexion.remove(interval[0])
                self.inflexion.remove(interval[1])
            else:
                #print 'dot_two'
                #print inflexion[inflexion.index(interval[1]) + 1]
                #print interval
                #print self.inflexion
                self.inflexion.remove(inflexion[inflexion.index(interval[1]) + 1])
                #print self.vertex      
                self.inflexion.remove(interval[1])


    def handleUpMax(self,interval,inflexion):         ###处理上涨波段最大值不是可疑顶点
        self.series = [1.0/item for item in self.series]
        self.handleFallMin(interval,inflexion)
        self.series = [1.0/item for item in self.series]  
    
    
    def findException(self,inflexion): ##找出异常并处理异常
        
        for i in range(1,len(inflexion)-1,1):
            if self.series[inflexion[i]] > self.series[inflexion[i-1]] and self.series[inflexion[i+1]] > self.series[inflexion[i]]:
                self.inflexion.remove(inflexion[i])
            elif self.series[inflexion[i]] < self.series[inflexion[i-1]] and self.series[inflexion[i+1]] < self.series[inflexion[i]]:
                self.inflexion.remove(inflexion[i])
            else:
                pass
             
        interval_list = [[self.inflexion[i],self.inflexion[i+1]] for i in range(len(self.inflexion)-1)]
        
        for interval in interval_list:
            if (max(self.series[interval[0]:interval[1]+1]) not in [self.series[interval[0]],self.series[interval[1]]]) and self.series[interval[0]] > self.series[interval[1]]: ## 下跌波段最大值不是可疑顶点
                #print 1,interval
                self.handleFallMax(interval,self.inflexion[:])  ##处理下跌波段最大值不是可疑顶点
                break
            elif (max(self.series[interval[0]:interval[1]+1]) not in [self.series[interval[0]],self.series[interval[1]]]) and self.series[interval[0]] < self.series[interval[1]]: ## 上涨波段最大值不是可疑顶点
                #print 2,interval
                self.handleUpMax(interval,self.inflexion[:])   ##处理上涨波段最大值不是可疑顶点
                break
            elif (min(self.series[interval[0]:interval[1]+1]) not in [self.series[interval[0]],self.series[interval[1]]]) and self.series[interval[0]] > self.series[interval[1]]: ## 下跌波段最小值不是可疑底点
                #print 3,interval
                self.handleFallMin(interval,self.inflexion[:]) ##处理下跌波段最小值不是可疑底点
                break
            elif (min(self.series[interval[0]:interval[1]+1]) not in [self.series[interval[0]],self.series[interval[1]]]) and self.series[interval[0]] < self.series[interval[1]]: ## 上涨波段最小值不是可疑底点
                #print 4,interval
                self.handleUpMin(interval,self.inflexion[:])  ##处理上涨波段最小值不是可疑底点
                break
            else:
                pass

    def isHandle(self):  ##判断异常波段是否完全被处理
        
        handle = False
        interval_list = [[self.inflexion[i],self.inflexion[i+1]] for i in range(len(self.inflexion)-1)]
        for interval in interval_list:
            
            if (max(self.series[interval[0]:interval[1]+1]) not in [self.series[interval[0]],self.series[interval[1]]]) or (min(self.series[interval[0]:interval[1]+1]) not in [self.series[interval[0]],self.series[interval[1]]]):
                #print interval
                handle = True
                break
        
        return handle
      
    def exceptionHandling(self):
        
        self.getInflexion()
        self.inflexion = self.vertex + self.nadir
        self.inflexion.sort()
        #print self.inflexion
        while(self.isHandle()):
            self.findException(self.inflexion[:])
```

```python
def wave_plot(series,inflexion):
    
    series['wave_band'] = series.iloc[inflexion]
    for i in range(len(series)):
        for j in range(len(inflexion)-1):
            if i <= inflexion[j+1] and i >= inflexion[j]:
                slope = (series['wave_band'].iloc[inflexion[j+1]] - series['wave_band'].iloc[inflexion[j]]) / (inflexion[j+1] - inflexion[j])
                series['wave_band'].iloc[i] = slope * (i-inflexion[j]) + series['wave_band'].iloc[inflexion[j]]
    
    series.plot(figsize= (20,9),linewidth=2)
```
