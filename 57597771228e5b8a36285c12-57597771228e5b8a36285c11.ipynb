* 既然提问者对平台不熟悉，我详细解释一下策略的实现。首先问题:"想写一个涨停开盘，盘中涨停裂开，最后重新涨停的T型打板策略，以涨停价买入，第二天开盘卖出。"严格来说这要写成分钟线策略，才不至于用到未来数据。但如果不用未来数据，你也不可能知道最后的日K是一个T型。

* OK,本策略主要是用了两个现成的API,DataAPI.MktEqudGet获取行情数据，DataAPI.MktLimitGet获取股票涨停次数及涨停价格。然后筛选出涨停次数大于1，且开盘价等于涨停价、收盘价等于涨停价的股票，等权买入次日卖出。实际上这个思路可以直接在代码模式下实现，并可以分析的更清楚。

* 总体来说，个人觉得这个是一个不错的指标啊，胜率很高，会进一步探索分钟线的可行性，感谢@quantking的提问。
