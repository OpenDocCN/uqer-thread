```python
import numpy as np
import pandas as pd
from pandas import DataFrame

import datetime

import sklearn
from sklearn.linear_model import LogisticRegression
from sklearn.lda import LDA
from sklearn.qda import QDA

import matplotlib.pyplot as plt

from CAL.PyCAL import *
cal = Calendar('China.SSE')
```

```python
#回测区间
start = '20060501'                       # 回测起始时间
end   = '20160501'

universe = [ '000001.XSHE']

#获取回测期间全部数据
bars = DataFrame()
days = 144
delta = '-'+str(days)+'B'
startDate=cal.advanceDate(start,delta,BizDayConvention.Preceding).strftime('%Y%m%d')

bars = DataAPI.MktEqudAdjGet(secID=universe[0],beginDate=startDate,endDate=end,field=u"tradeDate,openPrice,highestPrice,lowestPrice,closePrice,turnoverVol",pandas="1")
bars = bars.rename(columns={'tradeDate':'Date','openPrice':'Open','highestPrice':'High','lowestPrice':'Low','closePrice':'Close','turnoverVol':'Volume'})
bars['Date'] =  pd.to_datetime(bars['Date']) #因为DataAPI所有日期都是以字符串形式存储的，因此此处需要转换成日期时间型,这里很重要！
bars = bars.set_index('Date')#转换后index是DatetimeIndex型
bars.head()
```

```python
# Create the new lagged DataFrame
lags = 5
tslag = pd.DataFrame(index=bars.index)
tslag["Today"] = bars["Close"]
tslag["Volume"] = bars["Volume"]

# Create the shifted lag series of prior trading period close values
for i in xrange(0,lags):
    tslag["Lag%s" % str(i+1)] = bars["Close"].shift(i+1)

# Create the returns DataFrame
tsret = pd.DataFrame(index=tslag.index)
tsret["Volume"] = tslag["Volume"]
tsret["Today"] = tslag["Today"].pct_change()*100.0

# If any of the values of percentage returns equal zero, set them to
# a small number (stops issues with QDA model in scikit-learn)
for i,x in enumerate(tsret["Today"]):
    if (abs(x) < 0.0001):
        tsret["Today"][i] = 0.0001

# Create the lagged percentage returns columns
for i in xrange(0,lags):
    tsret["Lag%s" % str(i+1)] = tslag["Lag%s" % str(i+1)].pct_change()*100.0

# Create the "Direction" column (+1 or -1) indicating an up/down day
tsret["Direction"] = np.sign(tsret["Today"])
snpret = tsret[tsret.index >= start]
snpret.head(10)
```

```python
def fit_model(name, model, X_train, y_train, X_test, pred):
    """Fits a classification model (for our purposes this is LR, LDA and QDA)
    using the training data, then makes a prediction and subsequent "hit rate"
    for the test data."""

    # Fit and predict the model on the training, and then test, data
    model.fit(X_train, y_train)
    pred[name] = model.predict(X_test)

    # Create a series with 1 being correct direction, 0 being wrong
    # and then calculate the hit rate based on the actual direction
    pred["%s_Correct" % name] = (1.0+pred[name]*pred["Actual"])/2.0
    hit_rate = np.mean(pred["%s_Correct" % name])
    print "%s: %.3f" % (name, hit_rate)
```

```python
start_test = '20150501'
        
# Use the prior two days of returns as 
# predictor values, with direction as the response
X = snpret[["Lag1","Lag2"]]
y = snpret["Direction"]

# Create training and test sets
X_train = X[X.index < start_test]
X_test = X[X.index >= start_test]
y_train = y[y.index < start_test]
y_test = y[y.index >= start_test]


# Create prediction DataFrame
pred = pd.DataFrame(index=y_test.index)
pred["Actual"] = y_test

# Create and fit the three models    
print "Hit Rates:"
models = [("LR", LogisticRegression()), ("LDA", LDA()), ("QDA", QDA())]
for m in models:
    fit_model(m[0], m[1], X_train, y_train, X_test, pred)
```

```python
#产生信号

# Create the predicting factors for use 
# in direction forecasting
predictors = X[X.index >= start_test]

"""Returns the DataFrame of symbols containing the signals
to go long, short or hold (1, -1 or 0)."""
signals = pd.DataFrame(index=predictors.index)
signals['signal'] = 0.0       



# Create the Quadratic Discriminant Analysis model
# and the forecasting strategy
model = QDA()
model.fit(X_train, y_train)


signals['signal'] = model.predict(predictors)

# Remove the first five signal entries to eliminate
# NaN issues with the signals DataFrame
signals['signal'][0:5] = 0.0
signals['positions'] = signals['signal'].diff() 

signals[:25]
```

```python
#定义头寸
def generate_positions(signals):
    positions = pd.DataFrame(index=signals.index).fillna(0.0)
    symbol = universe[0]
    positions[symbol] = 500*signals['signal']  
    return positions
```

```python
#计算组合
def portfolio(initial_capital,positions,bars):
    portfolio = positions*bars['Close']
    pos_diff = positions.diff()

    portfolio['holdings'] = (positions*bars['Close']).sum(axis=1)
    portfolio['cash'] = initial_capital - (pos_diff*bars['Close']).sum(axis=1).cumsum()

    portfolio['total'] = portfolio['cash'] + portfolio['holdings']
    portfolio['returns'] = portfolio['total'].pct_change()
    return portfolio

initial_capital = 100000
positions = generate_positions(signals)
returns = portfolio(initial_capital,positions,bars)
returns.tail(10)
```

```python
# 图表化
fig = plt.figure(figsize=(18,12))
fig.patch.set_facecolor('white')

# Plot the price 
ax1 = fig.add_subplot(211,  ylabel=' price ')
bars['Close'].plot(ax=ax1, color='r', lw=2.)


# Plot the equity curve
ax2 = fig.add_subplot(212, ylabel='Portfolio value ')
returns['total'].plot(ax=ax2, lw=2.)

fig.show()
```

```python
#研究策略性能

#夏普率公式
def create_sharpe_ratio(returns, periods=250):
    """
    Create the Sharpe ratio for the strategy, based on a 
    benchmark of zero (i.e. no risk-free rate information).

    Parameters:
    returns - A pandas Series representing period percentage returns.
    periods - Daily (250), Hourly (250*4), Minutely(250*4*60) etc.
    """
    return np.sqrt(periods) * (np.mean(returns)) / np.std(returns)

#最大回撤公式
def create_drawdowns(equity_curve):
    """
    Calculate the largest peak-to-trough drawdown of the PnL curve
    as well as the duration of the drawdown. Requires that the 
    pnl_returns is a pandas Series.

    Parameters:
    pnl - A pandas Series representing period percentage returns.

    Returns:
    drawdown, duration - Highest peak-to-trough drawdown and duration.
    """

    # Calculate the cumulative returns curve 
    # and set up the High Water Mark
    # Then create the drawdown and duration series
    hwm = [0]
    eq_idx = equity_curve.index
    drawdown = pd.Series(index = eq_idx)
    duration = pd.Series(index = eq_idx)

    # Loop over the index range
    for t in range(1, len(eq_idx)):
        cur_hwm = max(hwm[t-1], equity_curve[t])
        hwm.append(cur_hwm)
        drawdown[t]= hwm[t] - equity_curve[t]
        duration[t]= 0 if drawdown[t] == 0 else duration[t-1] + 1
    return drawdown.max(), int(duration.max())
```

```python
strsr = create_sharpe_ratio(returns.returns,250)
print "Strategy Sharpe ratio: %s" %strsr
maxDrawdown,maxDrawdownDuration = create_drawdowns(returns.returns) # 最大回撤，最长回撤期
print "maxDrawdown:%s" %maxDrawdown
print "maxDrawdownDuration: %s" %maxDrawdownDuration
```
