```python
##取数据
import pandas as pd
import numpy as np
stock_list = ['000002', '601318', '600519', '600309', '002230']
df=DataAPI.MktEqudAdjAfGet(secID=u"",ticker=stock_list,tradeDate=u"",beginDate=u"20170101",endDate=u"20181231",isOpen="",field=u"tradeDate,ticker,closePrice",pandas="1").pivot_table(index='tradeDate',columns='ticker',values='closePrice')
df.rename(columns={})
df=df.pct_change().fillna(0)
df1=DataAPI.MktIdxdGet(ticker=u"399300",beginDate=u"20170101",endDate=u"20181231",exchangeCD=u"XSHE,XSHG",field=u"tradeDate,closeIndex",pandas="1").set_index('tradeDate').pct_change().fillna(0)
df1=df1.rename(columns={'closeIndex':'399300'})
df2=pd.concat([df,df1],axis=1)
##计算超额收益
rf=1.032**(1/360)-1
df_R=df2-rf

##回归实现
import statsmodels.api as sm
stock_names = {
 '000002': '万科A',
 '601318': '中国平安',
 '600519': '贵州茅台',
 '600309': '万华化学',
 '002230': '科大讯飞'
}
for stock in ['000002', '601318', '600519', '600309', '002230']:

    model = sm.OLS(df_R[stock], sm.add_constant(df_R['399300']))
result = model.fit()
print(stock_names[stock] + '')
print(result.summary())
print('')
```
