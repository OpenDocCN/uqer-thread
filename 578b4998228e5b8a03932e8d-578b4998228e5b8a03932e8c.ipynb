主要是指数编制的动态性和回测股票池静态性导致的问题。

然后在好奇心的驱使下就顺便测了一下：
1. 假如买入的股票从开始到最后都是在300指数里面。
2. 假如买入的股票开始在300指数里面， 但是后面被剔除出指数编制。
3. 假如买入股票时没有在300指数里面， 但是后来被收录了。

回测区间2010-01-01~2016-07-01, 沪深300的年化收益是-2%, 

回测采用的是等权买入， 结果：
1. 买入的股票从开始到最后都在300指数里面， 年化收益：11.7%
2. 买入的股票开始在300指数里面， 年化收益：-10.2%
3. 未仆先知，买入股票时没有在300指数里面， 但是后来被收录了， 年化收益： 34%

第三点的几乎没有实际操作指导作用。
第一点说明买入持有业务稳定，业绩稳定的百年老店可以长期跑赢300指数。
第二点说明， 300指数里面也有很多股票是没有投资价值的。
