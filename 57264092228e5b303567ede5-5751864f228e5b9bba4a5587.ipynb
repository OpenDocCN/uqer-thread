# 本杰明·格雷厄姆的价值投资十准则：根据A股市场特殊性的修改与校准

* **经典价值投资策略** 由格雷厄姆晚年在一次访谈中提出，被认为是格雷厄姆一生交易经验的最重要总结。

* 经典价值投资策略共由十条准则组成，这十条准则又可分为价值五法和安全五法。
* **价值五法** (前五条)
1. 股票的盈利回报率（市盈率倒数）大于美国AAA级债券收益的2倍。`这里以三年定期存款利率代替`
2. 股票的市盈率应小于其过去五年最高市盈率的40%。`考虑A股的情况，以20%代替`
3. 股票派息率应大于美国AAA级债券收益率的2/3。
4. 股价要低于每股有形资产净值的2/3。
5. 股价要低于每股净流动资产（流动资产 - 流动负债）的2/3。`回测发现此条在股价严重失真的A股几乎没有股票符合条件，直接删去不用`

* **安全五法** (后五条)
6. 总负债要小于有形资产净值。`根据初步回测结果，平衡起见将这个比值强化到0.8`
7. 流动比率（流动资产 / 流动负债）要大于2。 `回测发现几乎没有符合的标的，改用1`
8. 总负债要要小于净流动资产的2倍。 `回测发现几乎没有符合的标的，改用3`
9. 过去10年的平均年化盈利增长率要大于7%。 `数据库有限的缘故，改用有数据以来大于7%`
10.  过去十年中不能超过2次的盈利增长率小于-5%。`经多次尝试改用-10%，次数不超过2次`

* **注意**  市场、时期不同切莫生搬硬套，格雷厄姆自己也认为可以完全满足这十项准则的公司已越来越少（1976，美国），因此笔者认为应该根据A股具体情况适当调整，并且只要满足**安全五法**中三条即可买入。
