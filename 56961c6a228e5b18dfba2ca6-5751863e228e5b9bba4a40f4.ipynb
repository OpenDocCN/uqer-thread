# 【小散技术分析】--均线择时

- 均线一直是经验散户的主要择时手段之一，本文提供了一种基于5日均线和12日均线的择时方案，欢迎拍砖。

   ### 买入条件：
    - $1$：5日均线上穿12日均线，标记位置D，5日均线值记为d
    - $2$：此前最近的一次5日线上穿12日线，标记位置B，5日均线值记为b
    - $3$：B点和D点之间5日均线最小值，标记为C，5日均线值记为c
    - $4$：B点向前数5天，该区间内的5日均线最小值，标记为A，5日均线值记为a
    - $5$：原则上d>c>b>a为买点，实际操作中满足b<d 或 a<d 或 b<c 均可作为买入触发信号。
