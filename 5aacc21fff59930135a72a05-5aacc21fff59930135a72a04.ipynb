## 考夫曼自适应均线(KAMA)
说到考夫曼自适应均线得先说一下指数平滑移动平均线(EMA),相较于普通的移动平均线对过去数个交易日赋予均等权重而在市场快速运动时反应迟滞，EMA赋予最近交易日更高的权重，EMA = EMA[-1]+C*(close[-1]-EMA[-1]).
	假设10日指数平滑移动平均线赋予最后一天2/(10+1) = 18.18%的权重(C=2/11)， EMA = ((close[-1]-EMA[-1]) x weight + EMA[-1])，迭代后得到EMA = 1/2*close[-1] + (1/2)^2*close[-2] +…………+ (1/2)^n*close[-n],亦即过往价格对当日EMA的贡献呈指数型递减.
	虽然通过赋予附近交易日更大的权重很大程度地解决了均线的滞后性，不过简单这样给予递减权重还是有些草率。例如数个交易日前股票放量暴拉近几个交易日盘整，把大部分权重都给最近几个交易日是不合适的(之前暴涨体现不出来)。
    
考夫曼自适应曲线原理上也是一种指数平滑移动平均线，不过它的优点是可以自适应调节权重系数C，在股价快速移动时C较大均线反应速度较快，在股价盘整时C较小避免均线过度反应。这种功能光是听起来就屌屌的啊~ 比如之前几日的暴涨行情会得到比近几个盘整交易日更大的权重。
	考夫曼是通过引入一个叫效率系数(Efficiency Ratio)的因子来做到这一点的。效率系数等于价格改变(正值)/波动率，其中价格改变等于N个时间周期中价格的净变化，change =abs(price[-1] – price[-n]),波动性等于N个周期中所有单周期的价格变化的总和（每一个价格变化都使用正数）,Volatility = ∑|(close - prior close)|.效率系数ER在0到1之间波动，当市场处于单边行情时趋近于1，当市场盘整时效率系数趋近于0.系数C = [ER*(fastest - slowest)+ slowest]^2，考夫曼取短周期系数fastest=2/(2+1) = 2/3，长周期系数slowest=2/(30+1) = 2/31.

### KAMA均线计算步骤：
1.计算n个交易日收盘价改变值的绝对值Change = |(close[0] - close[-9]|
2.计算这n个交易日的波动率，作者使用近n日每日的收盘价减去前一日的收盘价的绝对值并加总这n-1个绝对值：Volatility = ∑|(close - prior close)|
3.计算效率系数ER = Change / Volatility
4.计算平滑常数C = [ER*(fastest - slowest)+ slowest]^2，当效率系数一直为1时前一日的收盘价占均线的权重为4/9，也就是自适应均线变成了2.25日线，当效率系数一直为0时前一日的收盘价占均线的权重为4/961，自适应均线变成了240日线
5.计算KAMA值，KAMA = KAMA[-1]+C*(close[-1]-KAMA[-1])

```python
from pandas import DataFrame,Series
import numpy as np
import pandas as pd
import random
from fractions import *
#universe = set_universe('HS300') #设置股票池为沪深300
#stock_pool = random.sample(universe, 50) #由于300支股票进行长时间回测太耗时跑不动，我们随机从里面选50支

def autoMovingAverage(stock,beginDate,endDate):
    
    df = DataAPI.MktEqudGet(secID=stock,beginDate=beginDate,endDate=endDate,isOpen="1",pandas="1")
    df.index = df.loc[:,('tradeDate')]
    close = df.loc[:]['closePrice']
    KAMA = [] #list用于储存KAMA均线的值
    for i in range(close.shape[0]-10):
        direction = close[i+9] - close[i] 
        change = abs(direction)
        volatility = 0
        for j in range(i,i+9):
            volatility = volatility + (abs(close[j+1]-close[j]))
        ER = change/volatility
        fastest = Fraction(2, 3)
        slowest = Fraction(2, 31)
        C = np.square(ER*(fastest - slowest) + slowest)
        if len(KAMA) == 0:  
            KAMA.append(np.mean(close[:10])) #KAMA的初始值取前10个交易日的平均值
        KAMA.append(KAMA[i] + C*(close[i+10] - KAMA[i])) 
        #前九个交易日均线没有数据，删除前九个交易日数据以对齐收盘价和均线
    return KAMA,close[9:]
```

```python
KAMA,close = autoMovingAverage('000100.XSHE','20160101','20171231') #时间太长不怎么看得清
Avg = Series(KAMA,index = close.index)
close.plot(figsize=(14, 7))
Avg.plot(figsize=(14, 7))
```

正如考夫曼所设计的那样，当股票陷入盘整时自适应均线上下波动幅度很小，显示出此时股票并没有向上或者向下的趋势；当股票启动/跳水时，自适应曲线的反应速度很快，紧紧跟上了股票走势
