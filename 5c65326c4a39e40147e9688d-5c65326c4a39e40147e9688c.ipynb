```python
import pandas as pd
datatmp=DataAPI.EquGet(secID=u"",ticker=u"",equTypeCD=u"A",listStatusCD=u"",field=u"exchangeCD,ticker,secShortName,totalShares",pandas="1")
#data1=data1.sort(["totalShares"],ascending=True)
#data1=data1.loc[data1["totalShares"]<150000000]
#print "data1=",data1
ticker=datatmp["ticker"].values
#print "stock:",stock
startdate="20181228"
enddate="20190131"
hkholdstart=DataAPI.HKshszHoldGet(tradeCD=u"",ticketCode=u"",secID=u"",ticker=ticker,partyName=u"",beginDate=startdate,endDate=startdate,field=u"",pandas="1")
hkholdend=DataAPI.HKshszHoldGet(tradeCD=u"",ticketCode=u"",secID=u"",ticker=ticker,partyName=u"",beginDate=enddate,endDate=enddate,field=u"",pandas="1")
#startPrice=DataAPI.MktEqudGet(secID=u"",ticker=ticker,beginDate=u"20181231",endDate=u"20181231",\
   #                           isOpen="", field=u"closePrice",pandas="1")
#print startPrice
start_Ma5=DataAPI.MktStockFactorsOneDayGet(tradeDate=enddate,secID=u"",ticker=ticker,field=u"ticker,tradeDate,MA5",pandas="1")
#print data_Ma5
end_Ma5=DataAPI.MktStockFactorsOneDayGet(tradeDate=startdate,secID=u"",ticker=ticker,field=u"ticker,tradeDate,MA5",pandas="1")
#print "end_Ma5:",end_Ma5
startdata=pd.merge(hkholdstart,start_Ma5)
enddata=pd.merge(hkholdend,end_Ma5)
#print "enddata:" ,enddata

#增加一列为partyVol*MA5/10000
startdata['amount']=startdata.apply(lambda x:x.partyVol*x.MA5/10000, axis = 1)
enddata['amount']=enddata.apply(lambda x:x.partyVol*x.MA5/10000, axis = 1)
startdata=startdata[['tradeDate','ticker','partyName','MA5','amount']]
enddata=enddata[['tradeDate','ticker','MA5','amount']]
#print "startdata:" ,startdata

#更改列名
startdata.rename(columns={"tradeDate":"starttradeDate","partyName":"partyName1","MA5":"startMA5","amount":"startamount"},inplace=True)
enddata.rename(columns={"tradeDate":"endtradeDate","MA5":"endMA5","amount":"endamount"},inplace=True)
result=pd.merge(startdata, enddata, on='ticker')
result['zengchijine']=result.apply(lambda x:x.endamount-x.startamount, axis = 1)
result['zengchibilv']=result.apply(lambda x:(x.endamount-x.startamount)/x.startamount, axis = 1)
result['zhanghu']=result.apply(lambda x:(x.endMA5-x.startMA5)/x.startMA5, axis = 1)
result=result.sort(["zengchibilv"],ascending=True)
#print result.tail(200)
#过滤掉一些奇异值
result=result.loc[result["zengchibilv"]<2]
print result.tail(200)
import matplotlib.pyplot as plt
# 中文标题有问题！！！
plt.title('zengchibilv-zhanghu', fontsize=24)
#https://www.cnblogs.com/hhh5460/p/5786115.html
#python 线性回归示例
from sklearn import linear_model
# 建立线性回归模型
regr = linear_model.LinearRegression()
# 拟合
regr.fit(result['zhanghu'].reshape(-1, 1), result['zengchibilv']) # 注意此处.reshape(-1, 1)，因为X是一维的！
# 不难得到直线的斜率、截距
a, b = regr.coef_, regr.intercept_
print "斜率:\n",a
# 1.真实的点
plt.scatter(result['zhanghu'], result['zengchibilv'], color='blue')
# 2.拟合的直线
plt.plot(result['zhanghu'], regr.predict(result['zhanghu'].reshape(-1,1)), color='red', linewidth=4)
#过滤掉一些脏数据
plt.xlim()
plt.ylim()
plt.axis()
#设置title和x,y抽的label
plt.xlabel("zengchibilv")
plt.ylabel("zhanghu")
#展示图片必须加
plt.show()
plt.savefig("zengchibilv-zhanghu")
#print result
result=result.tail(200)
result.to_csv("zengchijine.csv")
```
