```python
import pandas as pd 
import seaborn as sns
time = [pd.to_datetime(str(dt)).strftime('%Y-%m-%d') for dt in bt.tradeDate] 
price = []
for i in range(0,len(bt.security_position.values)):
    price.append(float(bt.security_position.values[i].values()[0]['value']))
dataframe = pd.DataFrame({'time':time,'price':price})
dataframe['time'] = pd.to_datetime(dataframe['time'])
dataframe.set_index('time',inplace='TRUE')
dataframe[u'无风险利率'] = (4.0 / 100 + 1) ** (1.0 / 250) - 1
dataframe[u'无风险收益_净值'] = (dataframe[u'无风险利率'] + 1).cumprod()
by_month = dataframe.resample('M', how='first', kind='period')
trade_log = pd.DataFrame(index=by_month.index)
trade_log['基金净值'] = by_month['price'] / 10000  # 以指数当天收盘点位除以1000作为单位基金净值
trade_log['money'] = 10000  # 每月月初投入1000元申购该指数基金
    # 定投购买余额宝等无风险产品
trade_log[u'理财份额'] = trade_log['money'] / by_month[u'无风险收益_净值']  # 当月的申购份额
trade_log[u'总理财份额'] = trade_log[u'理财份额'].cumsum()  # 累积申购份额
trade_log[u'累计定投资金'] = trade_log['money'].cumsum()
temp = trade_log.resample('D', fill_method='ffill')
dataframe = dataframe.to_period('D')
daily_data = pd.concat([dataframe, temp[[u'总理财份额', u'累计定投资金']]], axis=1, join='inner')
daily_data[u'理财定投资金曲线'] = daily_data[u'无风险收益_净值'] * daily_data[u'总理财份额']
daily_data[['price', u'理财定投资金曲线']].plot(grid=False, figsize=(12,8))
sns.despine()
```
