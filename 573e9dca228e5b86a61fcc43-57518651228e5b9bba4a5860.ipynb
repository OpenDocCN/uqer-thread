都说程序员最理性，每天面对的代码冷冰冰。NO！NO！NO！除了功与名，代码里其实深藏铁汉柔情！

520，1314，用代码说爱，一键克隆，向心爱的人表白吧！

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn

x = np.linspace(-3, 3, 50, endpoint=True)
y1= 1/(x+3)
y2 = np.sqrt(9-x*x)
y3 = abs(2*x)
y4 = -3*abs(np.sin(x))

plt.figure(figsize = (2,2))
plt.plot(x,y1)
plt.show()

plt.figure(figsize = (2,2))
plt.plot(x,y2,x,-y2)
plt.show()

plt.figure(figsize = (2,2))
plt.plot(x,y3)
plt.show()

plt.figure(figsize = (2,2))
plt.plot(y4,x)
plt.show()
```
