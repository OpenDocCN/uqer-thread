## 04. 自学成才第十二步——利用因子贡献度挑选因子进行机器学习——Jason的学习笔记 ##
地址：https://uqer.datayes.com/community/share/58427dfc228e5b49f14ad45e
算法：随机森林
因子：同上
思路：同上
特殊：
1：前两篇都是因子固定的，从优化/变换特征角度处理。此篇用来做特征筛选，特征的优劣评估。
因子贡献度的含义如下：
1.首先将股票池中的股票按因子进行排名，分别选出排名靠前的 20%和排名靠后的20%股票构成两个组合；
2.我们将两个组合的平均收益差和股票池中所有股票涨跌前后20%股票的平均收益差相除，得到的比值即为因子贡献度；
所以简答来说就是增加了一个特征筛选步骤。


## 05. 机器学习&量化投资-从入门到放弃全套笔记-LogisticRegression ## 
地址：https://uqer.datayes.com/community/share/5788af62228e5b8a03932cc5
算法：逻辑回归
因子：["LFLO","PB","PE","VOL10","VOL20","MA10","MA20", "RSI", "Volatility"]
思路：
调仓周期：10工作日
模型训练：
定义超额收益：一定时间t内个股相对大盘涨跌率,本文中为“个股10日涨跌率-大盘10日涨跌率”
从T-1日得到训练数据：T-1-10日的因子=〉T-1日的超额收益率，超额收益率二值化（上涨(大于0)则取1，下跌（小于0）则取0）
从T-2日得到训练数据：：T-2-10日的因子=〉T-2日的超额收益率，超额收益率二值化
，，，
从T-60日得到训练数据：：T-60-10日的因子=〉T-60日的超额收益率，超额收益率二值化
训练数据是非常充分的，取得的方法也比较特别，向前取了60天的。
模型预测：超额收益率（1涨0跌）
买卖策略：由于逻辑回归得到分类概率，取最大的top20，买卖规则为优矿默认的，卖出上周期，重新等权买入
特殊：
1：因子做了去极值，标准化处理
2：去特殊股票，st，流动性差，新上市等

# 小结 #
------
目前社区使用机器学习的算法，基本都是取因子截面
因子：涨跌率，成交量变动率，ma10等，以及pe，动量等可以从量价推算出的因子等
算法：随机森林较多，其次逻辑回归
注意：因子中的secid和tradedate在训练预测时都会去掉，不再单独注明

# 个人理解 #
-----
机器学习中算法基本都已经现成的了，也无需考虑在开发出全新算法（那是学术界的事，工业界讲究拿来主义，多快好省）
相关人工工作主要是数据清理和特征处理上以及后续调参调优上（体力活为主）。
但眼下优矿策略基本都是拿来特征直接使用，未做任何处理，这种原始数据直接用算法，很大概率是悲剧的。
后续计划尝试对价格信息进行抽象（目前想到的可以用向量模型[时间周期，涨跌率百分比]来代替原始k线进行趋势或价格转折点的预测，只是大概有个idea，未作太深入的考虑）
