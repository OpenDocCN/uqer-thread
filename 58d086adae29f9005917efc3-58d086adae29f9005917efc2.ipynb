# 价量二维时间序列聚类分析

```python
import datetime
import math
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
from CAL.PyCAL import *
cal = Calendar('China.SSE')
font.set_size(15)
todaydate=datetime.datetime.now().strftime('%Y-%m-%d')
lastTD=cal.advanceDate(todaydate,'-1B',BizDayConvention.Preceding).strftime('%Y-%m-%d')#last trading date
datah=DataAPI.MktEqudGet(secID=u"",ticker='000001'
                      ,isOpen=u"1",beginDate=u'1999-06-28',endDate=todaydate
                                ,field=u"secShortName,tradeDate,closePrice,turnoverVol,accumAdjFactor,chgPct"
                              ,pandas="1")
##复权处理
datah['closePrice']=datah['closePrice']*datah['accumAdjFactor']
datah['turnoverVol']=datah['turnoverVol']/datah['accumAdjFactor']
##initial critical points--local extrema
from scipy.signal import argrelextrema
locmax =  argrelextrema(datah['closePrice'].values,comparator=np.greater,order=20)
locmin =argrelextrema(datah['closePrice'].values, np.less,order=20)#return tuple
#initial fine-grain
loc=np.concatenate((list(locmax)[0],list(locmin)[0]),axis=0)
loc.sort()
```

```python
from sklearn import preprocessing
def mergecost(x,i,loc,p=1):
    if i==0:
        s1=x.iloc[:(loc[i+1]+1)]
    elif i<(len(loc)-1):
        s1=x.iloc[(loc[i-1]+1):(loc[i+1]+1)]  
    else:
        s1=x.iloc[(loc[i-1]+1):]  
        
    a=s1[['closePrice','turnoverVol']]
    a_scaled = preprocessing.scale(a)##scale and center
    U, s, V = np.linalg.svd(a_scaled, full_matrices=True)#SVD of 1st segment
    pro=np.dot(a_scaled,np.transpose(V[0:p,]))
    p=np.dot(pro,V[0:p,])#reconstruction by dimension 1
    #Q1=np.mean(np.square(p))
    error=a_scaled-p
    Q2=np.mean(np.diagonal(np.dot(error,np.transpose(error))))
    return Q2    

def bottomup(dat,seg_num,initial_loc,p=1):
    loc = initial_loc
    mc=[mergecost(dat,i,loc=loc,p=p) for i in range(len(loc))]
    while len(loc) > seg_num:
        j=mc.index(min(mc))#seg to merge
        loc=np.delete(loc,j)#update loc
        mc=np.ndarray.tolist(np.delete(mc,j))
        
        if j==0:#update mc
            mc[0]=mergecost(dat,0,loc=loc,p=p)
        elif j<len(loc):    
            mc[j-1]=mergecost(dat,j-1,loc=loc,p=p)
            mc[j]=mergecost(dat,j,loc=loc,p=p)
        else:
            mc[j-1]=mergecost(dat,j-1,loc=loc,p=p)
    return loc

loc_final=bottomup(datah,50,loc,1)
loc_final
```

```python
##cluster--object
def segmentClusterObject(dat,seglocs,p):
    seglocs = np.append(seglocs,dat.shape[0])
    obj=np.ones((len(seglocs),2))
    for j,l in enumerate(seglocs):
        if j==0:
            s1 = dat.iloc[0:(l+1)] 
        else:
            s1 = dat.iloc[(seglocs[j-1]+1):(l+1)] 
            
        a=s1[['closePrice','turnoverVol']]
        a_scaled = preprocessing.scale(a)##scale and center
        U, s, V = np.linalg.svd(a_scaled, full_matrices=True)#SVD of 1st segment
        S=np.diag(s)
        if p==1:
            object=s[0]*V[0,]
        else:
            object=np.dot(S[0:p,0:p],np.transpose(V[0:p,]))
        obj[j,] = object
    return obj

clusterObj=segmentClusterObject(datah,loc_final,1)
```

```python
import random
from scipy.spatial.distance import cdist 

def kmeans( X, centres, delta=.001, maxiter=10, metric="euclidean", p=2, verbose=1 ):
    X = np.asanyarray(X)  
    centres = centres.copy()
    N, dim = X.shape
    k, cdim = centres.shape
    if dim != cdim:
        raise ValueError( "kmeans: X %s and centres %s must have the same number of columns" % (
            X.shape, centres.shape ))
    if verbose:
        print "kmeans: X %s  centres %s  delta=%.2g  maxiter=%d  metric=%s" % (
            X.shape, centres.shape, delta, maxiter, metric)
    allx = np.arange(N)
    prevdist = 0
    for jiter in range( 1, maxiter+1 ):
        D = cdist( X, centres, metric=metric, p=p )  # |X| x |centres|
        xtoc = D.argmax(axis=1)  # X -> nearest centre
        distances = D[allx,xtoc]
        avdist = distances.mean()  # median ?
        if verbose >= 2:
            print "kmeans: av |X - nearest centre| = %.4g" % avdist
        if (1 - delta) * prevdist <= avdist <= prevdist \
        or jiter == maxiter:
            break
        prevdist = avdist
        for jc in range(k):  # (1 pass in C)
            c = np.where( xtoc == jc )[0]
            if len(c) > 0:
                centres[jc] = X[c].mean( axis=0 )
                
    if verbose:
        print "kmeans: %d iterations  cluster sizes:" % jiter, np.bincount(xtoc)
    if verbose >= 2:
        r50 = np.zeros(k)
        r90 = np.zeros(k)
        for j in range(k):
            dist = distances[ xtoc == j ]
            if len(dist) > 0:
                r50[j], r90[j] = np.percentile( dist, (50, 90) )
        print "kmeans: cluster 50 % radius", r50.astype(int)
        print "kmeans: cluster 90 % radius", r90.astype(int)
            # scale L1 / dim, L2 / sqrt(dim) ?
    return centres, xtoc, distances

def kmeanssample( X, k, nsample=0, **kwargs ):
    """ 
        1) kmeans a random sample of nsample ~ sqrt(N) from X
        2) full kmeans, starting from those centres
    """
    N, dim = X.shape
    if nsample == 0:
        nsample = max( 2*np.sqrt(N), 10*k )
    Xsample = randomsample( X, int(nsample) )
    pass1centres = randomsample( X, int(k) )
    samplecentres = kmeans( Xsample, pass1centres, **kwargs )[0]
    return kmeans( X, samplecentres, **kwargs )

def randomsample( X, n ):
    sampleix = random.sample( xrange( X.shape[0] ), int(n) )
    return X[sampleix]

def nearestcentres( X, centres, metric="euclidean", p=2 ):
    D = cdist( X, centres, metric=metric, p=p )  # |X| x |centres|
    return D.argmin(axis=1)

def EignVecsMatrixSimilar(x,y):
    """
    similarity is defined as the sum/mean of the square of the cosines(inner product) between each vector of matrix
    """
    return np.inner(np.inner(x,y),np.inner(y,x))

#...............................................................................
class Kmeans:
    def __init__( self, X, k=0, centres=None, nsample=0, **kwargs ):
        self.X = X
        if centres is None:
            self.centres, self.Xtocentre, self.distances = kmeanssample(
                X, k=k, nsample=nsample, **kwargs )
        else:
            self.centres, self.Xtocentre, self.distances = kmeans(
                X, centres, **kwargs )

    def __iter__(self):
        for jc in range(len(self.centres)):
            yield jc, (self.Xtocentre == jc)
```
