## 第一步：获取优矿因子列表

```python
import matplotlib.pyplot as plt
from matplotlib import rc
from matplotlib import dates
rc('mathtext', default='regular')
import seaborn as sns
sns.set_style('white')
import datetime
import numpy as np
import pandas as pd
import time
import scipy.stats as st
from CAL.PyCAL import * 

## 获取优矿因子列表
begin_date = '20170101'  # 开始日期
end_end = '20170401'    # 结束日期
columns=DataAPI.MktStockFactorsOneDayGet(tradeDate=u"20150227",secID=u"",ticker=u"000001",field=u"",pandas="1").columns
all_factors=columns.tolist()
factor_list = all_factors
factor_list.remove('secID')
factor_list.remove('ticker')
factor_list.remove('tradeDate')
print factor_list
```

## 第三步：使用优矿的参数回测框架进行factor调优尝试

```python
# 把回测参数封装到 SimulationParameters 中，供 quick_backtest 使用
max_history_window=100
sim_params = quartz.SimulationParameters(start, end, benchmark, universe, capital_base, refresh_rate=refresh_rate, max_history_window=max_history_window)

# ---------------回测参数部分结束----------------
# 获取回测行情数据
data = quartz.get_backtest_data(sim_params)
# 运行结果
results = pd.DataFrame(columns=['alpha','sharpe','max_drawdown'])
# 调整参数快速回测
for factor in (factor_list):
    start_time=time.time()
    try:
        #因子最小的10%和最大的10%都做尝试
        for asc_method in [True,False]:
            # 生成策略对象
            strategy = quartz.TradingStrategy(initialize, handle_data)
            # -------------------策略定义结束--------------------   
            # 开始回测
            bt, perf = quartz.quick_backtest(sim_params, strategy, data=data)

            # 保存运行结果
            field_name=factor+'_'+str(1 if asc_method==True else 0)
            results.loc[field_name]=pd.Series({'max_drawdown': perf['max_drawdown'],
                              'sharpe': perf['sharpe'],
                              'alpha': perf['alpha']
                             })
    except:
        print 'factor except'+factor
    finish_time = time.time()
    print 'factor:'+factor+','+str(finish_time-start_time)
results.tail()        
```
