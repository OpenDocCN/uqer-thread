# 2、策略原理
  策略的原理非常简单：  
  1. 用 MACD 和 CCI 两个指标对50ETF(或者50指数，这个我没有试过)进行择时，即用这两个指标最常见的买卖信号；
  2. 两个指标都看多的时候，卖出深度虚值的看跌期权；都看空的时候，卖出深度虚值的看涨期权；两个指标意见不和的时候，看震荡，双向卖出两份虚值期权。

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ETF_Data = DataAPI.MktFunddGet(tradeDate=u"",secID=u"",ticker=u"510050",beginDate=u"2015-02-10",endDate=u"2017-09-05",field=u"",pandas="1")
ETF_Data = ETF_Data[['tradeDate','openPrice','highestPrice','lowestPrice','closePrice','preClosePrice']]
```

```python
import talib
DIF, DEA, SIG = talib.MACD(ETF_Data['closePrice'].values, fastperiod=12, slowperiod=26, signalperiod=9)
CCI = talib.CCI(ETF_Data['highestPrice'].values, ETF_Data['lowestPrice'].values, ETF_Data['closePrice'].values, timeperiod=14)
```

```python
ETF_Data['DIF'] = DIF
ETF_Data['DEA'] = DEA
ETF_Data['CCI'] = CCI
```

```python
ETF_Data['MACD_BKsig'] = ETF_Data['DIF'].values>ETF_Data['DEA'].values
ETF_Data['CCI_BKsig'] = ETF_Data['CCI'].values > 100

ETF_Data['MACD_SKsig'] = ETF_Data['DIF'].values<ETF_Data['DEA'].values
ETF_Data['CCI_SKsig'] = ETF_Data['CCI'].values < 100
```

```python
ETF_Data = ETF_Data.set_index('tradeDate')
```

```python
ETF_Data['ETF_Daily_PL'] = np.zeros_like(ETF_Data['CCI'])
ETF_Data['Option_Daily_PL'] = np.zeros_like(ETF_Data['CCI'])
```

```python
# 初始化仓位信息
ETF_Position = 0 
Option_Position_C = 0
Option_Position_P = 0
for i in range(len(ETF_Data.index)-1):
    #根据昨日数据计算今日操作
    Long_sig = ETF_Data.ix[ETF_Data.index[i],'MACD_BKsig'] and ETF_Data.ix[ETF_Data.index[i],'CCI_BKsig']
    Short_sig = ETF_Data.ix[ETF_Data.index[i],'MACD_SKsig'] and ETF_Data.ix[ETF_Data.index[i],'CCI_SKsig'] 
    LS_sig = (ETF_Data.ix[ETF_Data.index[i],'MACD_BKsig'] and ETF_Data.ix[ETF_Data.index[i],'CCI_SKsig']) or (ETF_Data.ix[ETF_Data.index[i],'CCI_BKsig'] and ETF_Data.ix[ETF_Data.index[i],'MACD_SKsig'])
    #如果是多头信号，买入ETF,卖出虚值看跌期权
    if Long_sig and ETF_Position == 0 :
        #ETF的操作
        ETF_Position = 1
        ETF_BuyPrice = ETF_Data.ix[ETF_Data.index[i+1],'openPrice']
        ETF_Data.ix[ETF_Data.index[i+1],'ETF_Daily_PL'] = ETF_Data.ix[ETF_Data.index[i+1],'closePrice']-ETF_BuyPrice
        #期权操作：卖出虚值看跌期权(选择期限最长，虚值最深的)
        Option_Position_P = 1
        Option_Data =DataAPI.OptDpoGet(secID=u"",optID=u"",ticker=u"",varSecID=u"",varTicker=u"510050",varType=u"",beginDate=ETF_Data.index[i],endDate=ETF_Data.index[i],exerType=u"",contractType=u"",field=u"",pandas="1")
        Option_Data = Option_Data[['secID','strikePrice']]
        Option_Data['PorC'] = map(lambda x : x[6], Option_Data['secID'])        
        Option_Data['deliDate'] = map(lambda x : int(x[7:11]), Option_Data['secID'])
        Option_Data = Option_Data[Option_Data['PorC'] == 'P']
        Option_Data = Option_Data[Option_Data['deliDate'] == max(Option_Data['deliDate'])]
        Select_Opt = Option_Data[Option_Data['strikePrice'] == min(Option_Data['strikePrice'])]
        Select_SecID_P = Select_Opt['secID'].values[0]
        Select_Opt_Data = DataAPI.MktOptdGet(tradeDate=ETF_Data.index[i+1],secID=Select_SecID_P,optID=u"",ticker=u"",beginDate=u"",endDate=u"",field=u"",pandas="1")
        Sell_Price = Select_Opt_Data['openPrice'].values[0]
        if Option_Position_C == 1:
            Option_Position_C = 0 
            Opt_Data = DataAPI.MktOptdGet(tradeDate=ETF_Data.index[i+1],secID=Select_SecID_C,optID=u"",ticker=u"",beginDate=u"",endDate=u"",field=u"",pandas="1")
            BuyPrice = Opt_Data['openPrice'].values[0]
            PL = Opt_Data['preClosePrice'].values[0] - BuyPrice
        else:
            PL = 0
        ETF_Data.ix[ETF_Data.index[i+1],'Option_Daily_PL'] = Sell_Price - Select_Opt_Data['closePrice'].values[0]+PL
    if Short_sig:
        #期权操作：卖出虚值看涨期权(选择期限最长，虚值最深的)
        Option_Position_C = 1
        Option_Data =DataAPI.OptDpoGet(secID=u"",optID=u"",ticker=u"",varSecID=u"",varTicker=u"510050",varType=u"",beginDate=ETF_Data.index[i],endDate=ETF_Data.index[i],exerType=u"",contractType=u"",field=u"",pandas="1")
        Option_Data = Option_Data[['secID','strikePrice']]
        Option_Data['PorC'] = map(lambda x : x[6], Option_Data['secID'])        
        Option_Data['deliDate'] = map(lambda x : int(x[7:11]), Option_Data['secID'])
        Option_Data = Option_Data[Option_Data['PorC'] == 'C']
        Option_Data = Option_Data[Option_Data['deliDate'] == max(Option_Data['deliDate'])]
        Select_Opt = Option_Data[Option_Data['strikePrice'] == max(Option_Data['strikePrice'])]
        Select_SecID_C = Select_Opt['secID'].values[0]
        Select_Opt_Data = DataAPI.MktOptdGet(tradeDate=ETF_Data.index[i+1],secID=Select_SecID_C,optID=u"",ticker=u"",beginDate=u"",endDate=u"",field=u"",pandas="1")
        Sell_Price = Select_Opt_Data['openPrice'].values[0]
        if Option_Position_P == 1:
            Option_Position_P = 0 
            Opt_Data = DataAPI.MktOptdGet(tradeDate=ETF_Data.index[i+1],secID=Select_SecID_P,optID=u"",ticker=u"",beginDate=u"",endDate=u"",field=u"",pandas="1")
            BuyPrice = Opt_Data['openPrice'].values[0]
            PL = Opt_Data['preClosePrice'].values[0] - BuyPrice
        else:
            PL = 0
        ETF_Data.ix[ETF_Data.index[i+1],'Option_Daily_PL'] = Sell_Price - Select_Opt_Data['closePrice'].values[0]+PL
    if (Short_sig or LS_sig) and ETF_Position == 1:
        #ETF的操作
        ETF_Position = 0
        ETF_SellPrice = ETF_Data.ix[ETF_Data.index[i+1],'openPrice']
        ETF_Data.ix[ETF_Data.index[i+1],'ETF_Daily_PL'] = ETF_SellPrice-ETF_Data.ix[ETF_Data.index[i+1],'preClosePrice']
    if ETF_Position == 1:
        ETF_Data.ix[ETF_Data.index[i+1],'ETF_Daily_PL'] = ETF_Data.ix[ETF_Data.index[i+1],'closePrice']-ETF_Data.ix[ETF_Data.index[i+1],'preClosePrice']
    #震荡市期权双向卖出
    if LS_sig:
        if Option_Position_P ==1 and Option_Position_C ==1:
            pass
        if Option_Position_P ==1 and Option_Position_C ==0:
            Option_Position_C = 1
            Option_Data =DataAPI.OptDpoGet(secID=u"",optID=u"",ticker=u"",varSecID=u"",varTicker=u"510050",varType=u"",beginDate=ETF_Data.index[i],endDate=ETF_Data.index[i],exerType=u"",contractType=u"",field=u"",pandas="1")
            Option_Data = Option_Data[['secID','strikePrice']]
            Option_Data['PorC'] = map(lambda x : x[6], Option_Data['secID'])        
            Option_Data['deliDate'] = map(lambda x : int(x[7:11]), Option_Data['secID'])
            Option_Data = Option_Data[Option_Data['PorC'] == 'C']
            Option_Data = Option_Data[Option_Data['deliDate'] == max(Option_Data['deliDate'])]
            Select_Opt = Option_Data[Option_Data['strikePrice'] == max(Option_Data['strikePrice'])]
            Select_SecID_C = Select_Opt['secID'].values[0]
            Select_Opt_Data = DataAPI.MktOptdGet(tradeDate=ETF_Data.index[i+1],secID=Select_SecID_C,optID=u"",ticker=u"",beginDate=u"",endDate=u"",field=u"",pandas="1")
            Sell_Price = Select_Opt_Data['openPrice'].values[0]  
            Opt_Data = DataAPI.MktOptdGet(tradeDate=ETF_Data.index[i+1],secID=Select_SecID_P,optID=u"",ticker=u"",beginDate=u"",endDate=u"",field=u"",pandas="1")
            ETF_Data.ix[ETF_Data.index[i+1],'Option_Daily_PL'] =Sell_Price - Select_Opt_Data['closePrice'].values[0]+Opt_Data['preClosePrice'].values[0]-Opt_Data['closePrice'].values[0]
        if Option_Position_P ==0 and Option_Position_C ==1:
            Option_Position_P = 1
            Option_Data =DataAPI.OptDpoGet(secID=u"",optID=u"",ticker=u"",varSecID=u"",varTicker=u"510050",varType=u"",beginDate=ETF_Data.index[i],endDate=ETF_Data.index[i],exerType=u"",contractType=u"",field=u"",pandas="1")
            Option_Data = Option_Data[['secID','strikePrice']]
            Option_Data['PorC'] = map(lambda x : x[6], Option_Data['secID'])        
            Option_Data['deliDate'] = map(lambda x : int(x[7:11]), Option_Data['secID'])
            Option_Data = Option_Data[Option_Data['PorC'] == 'P']
            Option_Data = Option_Data[Option_Data['deliDate'] == max(Option_Data['deliDate'])]
            Select_Opt = Option_Data[Option_Data['strikePrice'] == min(Option_Data['strikePrice'])]
            Select_SecID_P = Select_Opt['secID'].values[0]
            Select_Opt_Data = DataAPI.MktOptdGet(tradeDate=ETF_Data.index[i+1],secID=Select_SecID_P,optID=u"",ticker=u"",beginDate=u"",endDate=u"",field=u"",pandas="1")
            Sell_Price = Select_Opt_Data['openPrice'].values[0]  
            Opt_Data = DataAPI.MktOptdGet(tradeDate=ETF_Data.index[i+1],secID=Select_SecID_C,optID=u"",ticker=u"",beginDate=u"",endDate=u"",field=u"",pandas="1")
            ETF_Data.ix[ETF_Data.index[i+1],'Option_Daily_PL'] =Sell_Price - Select_Opt_Data['closePrice'].values[0]+Opt_Data['preClosePrice'].values[0]-Opt_Data['closePrice'].values[0]  
        if Option_Position_P ==0 and Option_Position_C ==0:
            Option_Position_P = 1
            Option_Position_C =1
            Option_Data =DataAPI.OptDpoGet(secID=u"",optID=u"",ticker=u"",varSecID=u"",varTicker=u"510050",varType=u"",beginDate=ETF_Data.index[i],endDate=ETF_Data.index[i],exerType=u"",contractType=u"",field=u"",pandas="1")
            Option_Data = Option_Data[['secID','strikePrice']]
            Option_Data['PorC'] = map(lambda x : x[6], Option_Data['secID'])        
            Option_Data['deliDate'] = map(lambda x : int(x[7:11]), Option_Data['secID'])
            Option_Data_P = Option_Data[Option_Data['PorC'] == 'P']
            Option_Data_C = Option_Data[Option_Data['PorC'] == 'C']
            Option_Data_P = Option_Data_P[Option_Data['deliDate'] == max(Option_Data['deliDate'])]
            Option_Data_C = Option_Data_C[Option_Data['deliDate'] == max(Option_Data['deliDate'])]
            Select_Opt_P = Option_Data_P[Option_Data['strikePrice'] == min(Option_Data['strikePrice'])]
            Select_Opt_C = Option_Data_C[Option_Data['strikePrice'] == max(Option_Data['strikePrice'])]
            Select_SecID_P = Select_Opt_P['secID'].values[0]
            Select_SecID_C = Select_Opt_C['secID'].values[0]
            Select_Opt_Data_P = DataAPI.MktOptdGet(tradeDate=ETF_Data.index[i+1],secID=Select_SecID_P,optID=u"",ticker=u"",beginDate=u"",endDate=u"",field=u"",pandas="1")
            Sell_Price_P = Select_Opt_Data_P['openPrice'].values[0]  
            Select_Opt_Data_C = DataAPI.MktOptdGet(tradeDate=ETF_Data.index[i+1],secID=Select_SecID_C,optID=u"",ticker=u"",beginDate=u"",endDate=u"",field=u"",pandas="1")
            Sell_Price_C = Select_Opt_Data_C['openPrice'].values[0]

            ETF_Data.ix[ETF_Data.index[i+1],'Option_Daily_PL'] =Sell_Price_P - Select_Opt_Data_P['closePrice'].values[0]+Sell_Price_C - Select_Opt_Data_C['closePrice'].values[0] 
            
            
    if Option_Position_P ==1 and Option_Position_C ==1:
        Opt_Data_P = DataAPI.MktOptdGet(tradeDate=ETF_Data.index[i+1],secID=Select_SecID_P,optID=u"",ticker=u"",beginDate=u"",endDate=u"",field=u"",pandas="1")
        Opt_Data_C = DataAPI.MktOptdGet(tradeDate=ETF_Data.index[i+1],secID=Select_SecID_C,optID=u"",ticker=u"",beginDate=u"",endDate=u"",field=u"",pandas="1")
        ETF_Data.ix[ETF_Data.index[i+1],'Option_Daily_PL'] =Opt_Data_C['preClosePrice'].values[0]-Opt_Data_C['closePrice'].values[0] + Opt_Data_P['preClosePrice'].values[0]-Opt_Data_P['closePrice'].values[0]  
    if Option_Position_P ==0 and Option_Position_C ==1:
        #Opt_Data_P = DataAPI.MktOptdGet(tradeDate=ETF_Data.index[i+1],secID=Select_SecID_P,optID=u"",ticker=u"",beginDate=u"",endDate=u"",field=u"",pandas="1")
        Opt_Data_C = DataAPI.MktOptdGet(tradeDate=ETF_Data.index[i+1],secID=Select_SecID_C,optID=u"",ticker=u"",beginDate=u"",endDate=u"",field=u"",pandas="1")
        ETF_Data.ix[ETF_Data.index[i+1],'Option_Daily_PL'] =Opt_Data_C['preClosePrice'].values[0]-Opt_Data_C['closePrice'].values[0]# + Opt_Data_P['preClosePrice'].values[0]-Opt_Data_P['closePrice'].values[0]  
    if Option_Position_P ==1 and Option_Position_C ==0:
        Opt_Data_P = DataAPI.MktOptdGet(tradeDate=ETF_Data.index[i+1],secID=Select_SecID_P,optID=u"",ticker=u"",beginDate=u"",endDate=u"",field=u"",pandas="1")
        #Opt_Data_C = DataAPI.MktOptdGet(tradeDate=ETF_Data.index[i+1],secID=Select_SecID_C,optID=u"",ticker=u"",beginDate=u"",endDate=u"",field=u"",pandas="1")
        ETF_Data.ix[ETF_Data.index[i+1],'Option_Daily_PL'] =Opt_Data_P['preClosePrice'].values[0]-Opt_Data_P['closePrice'].values[0] #+ Opt_Data_P['preClosePrice'].values[0]-Opt_Data_P['closePrice'].values[0]  
    #print i
```

```python
option = ETF_Data['Option_Daily_PL']*10000
option = pd.DataFrame(option)
option.values[0] = 50000
option['Pure_Option_netvalue'] = option.values.cumsum()/50000
option = option.drop(['Option_Daily_PL'],axis=1)
option.plot(figsize=(10,7))
plt.grid()
```

```python
combine = (ETF_Data['ETF_Daily_PL']+ETF_Data['Option_Daily_PL'])*10000
combine = pd.DataFrame(combine)
combine.values[0] = 100000
combine['netvalue'] = combine.values.cumsum()/100000
combine = combine.drop(0,axis=1)
combine.plot(figsize=(10,7))
plt.grid()
```
