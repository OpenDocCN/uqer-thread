```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from datetime import timedelta
from statsmodels import regression
```

```python
#设置时间
beginDate = '20190104' 
endDate = '20190331'

#设置指数类型为上证50ETF
indexLabel = '510050' 

#读取指数
indexPrice = DataAPI.MktFunddGet(ticker=indexLabel,beginDate=beginDate,endDate=endDate,field=["tradeDate","closePrice"],pandas="1")
price = np.array(indexPrice.loc[:,'closePrice'])
#计算对数收益
back_price = np.append(price[0],price.copy())   
back_price = back_price[:-1] 
return_price = np.log(price) - np.log(back_price)
#计算收益率AR(1)的残差序列
r2 = back_price[1:]
r1 = back_price[:-1]
R = regression.linear_model.OLS(r2,r1).fit()
X = R.resid             
```

```python
#X    #需要处理的时间序列
#T    #时间序列的总长度（56）
#A    #子序列个数，
#n    #子序列长度（10-16）
#a    #子序列个数
#RS   R/S指数
#V    指数

RS = np.zeros(16)      #array
RS1 = np.zeros(16)      #array
RS2 = np.zeros(16)      #array
V = np.zeros(16)       #array
logRS = np.zeros(16)   #array
logn = np.zeros(16)   #array
X = np.array(X)    #读入时间序列,一维矩阵
T = X.shape[0]     #读取时间序列的总长度，
for n in xrange(10,16):  #每个长度计算一次
    a = int(T/n) 
    x1 = X[0:n*a].reshape(n,a)   #正向分段,n行a列，每一列为一个子序列
    x2 = X[(T - a * n): T].reshape(n,a)  #反向分段,n行a列，每一列为一个子序列
    m1 = np.mean(x1,axis=0)     #按列取均值，1行a列
    m2 = np.mean(x2,axis=0)     #按列取均值，1行a列
    p = np.ones((n,1))              #n行，1列
    y1 = x1 - p * m1                  #n行，a列，对每一列求离差
    y2 = x2 - p * m2                  #n行，a列，对每一列求离差       
    sig1 = np.std(x1,axis=0)             #1行，a列，对每一列求标准差
    sig2 = np.std(x2,axis=0)             #1行，a列，对每一列求标准差
    sum1 = np.cumsum(y1,axis=0)          #n行，a列，求累计离差
    sum2 = np.cumsum(y2,axis=0)          #n行，a列，求累计离差
    r1 = np.max(sum1,axis=0) - np.min(sum1,axis=0)      #%1行，a列
    r2 = np.max(sum2,axis=0) - np.min(sum2,axis=0)      #%1行，a列        
    RS1[n] = np.mean(r1 / sig1,axis=0)        #1行，1列
    RS2[n] = np.mean(r2 / sig2,axis=0)        #1行，1列    
    RS[n] = 0.5*RS1[n] + 0.5*RS2[n]
    logRS[n] = np.log(RS[n])
    V[n] = RS[n] * (pow(n,-0.5))
    logn[n] = np.log(n)
R = regression.linear_model.OLS(logRS[10:16],logn[10:16]).fit()
R.params[0]
```

```python
V统计量在13的时候出现最大值，姑且可以认为上证50ETF有一个平均循环周期，长度为13个交易日，所以我们接下来进行滚动Hurst指数计算的时候，滚动窗口定为13日
```

```python
def Hurst(XX,T):  
    XX = np.array(XX)    #读入时间序列,一维矩阵
    Lenth = XX.shape[0]     #读取时间序列的总长度，
    hurst = np.zeros(Lenth)
    for i in xrange(T,Lenth):
        X=XX[i-T:i+1]
        RS = np.zeros(T)      #array
        logRS = np.zeros(T)   #array
        logn = np.zeros(T)   #array        
        for n in xrange(10,T):  #每个长度计算一次
            a = int(T/n) 
            x1 = X[0:n*a].reshape(n,a)   #正向分段,n行a列，每一列为一个子序列
            x2 = X[(T - a * n): T].reshape(n,a)  #反向分段,n行a列，每一列为一个子序列
            m1 = np.mean(x1,axis=0)     #按列取均值，1行a列
            m2 = np.mean(x2,axis=0)     #按列取均值，1行a列
            p = np.ones((n,1))              #n行，1列
            y1 = x1 - p * m1                  #n行，a列，对每一列求离差
            y2 = x2 - p * m2                  #n行，a列，对每一列求离差       
            sig1 = np.std(x1,axis=0)             #1行，a列，对每一列求标准差
            sig2 = np.std(x2,axis=0)             #1行，a列，对每一列求标准差
            sum1 = np.cumsum(y1,axis=0)          #n行，a列，求累计离差
            sum2 = np.cumsum(y2,axis=0)          #n行，a列，求累计离差
            r1 = np.max(sum1,axis=0) - np.min(sum1,axis=0)      #%1行，a列
            r2 = np.max(sum2,axis=0) - np.min(sum2,axis=0)      #%1行，a列        
            RS1[n] = np.mean(r1 / sig1,axis=0)        #1行，1列
            RS2[n] = np.mean(r2 / sig2,axis=0)        #1行，1列    
            RS[n] = 0.5*RS1[n] + 0.5*RS2[n]
            logRS[n] = np.log(RS[n])
            logn[n] = np.log(n)
        R = regression.linear_model.OLS(logRS[10:T],logn[10:T]).fit()
        hurst[i] = R.params[0]
    return hurst
```

```python
时间窗口取为13，从第13个交易日开始，每个交易日都计算出一个Hurst指数的值，这个值是以该交易日和之前的13个交易日为基础算出来的。
```

```python
hurst = Hurst(X,T=13)
plt.plot(hurst)
```
