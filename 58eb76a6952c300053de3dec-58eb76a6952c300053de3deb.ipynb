有的策略在牛市表现突出，熊市则不佳。还有些在震荡市(猴)表现突出，牛市熊市则不佳。（牛熊猴通吃的应该极少，大部分策略其实都是有风格偏好，回测时间短的话可能无法体现）
为了方便策略进行风格判断，感觉有必要尝试下搞个风格标识模块。
如果目前市场不处于策略所擅长的区间，则降低主动股票仓位，减少回撤。

## 几点个人感受 
**1，风格判断具有延迟性。不要希望可以在牛市/熊市转换的那一天，恰好捕捉到信号，那样的话就是预测未来了，目前感觉不大可能。能做的仅仅是依靠历史信息，尽可能缩短预测延迟。

3，牛市时，大部分股票都是创x天新高。熊市反之
4，如果能判断牛熊，那么剩下就是猴市了，猴市本身特点是指数（或大多股票）都处于箱体内

```python
#5,20日均线
index_data['MA5']=pd.rolling_mean(index_data['closeIndex'],window=5)
index_data['MA10']=pd.rolling_mean(index_data['closeIndex'],window=10)
index_data['MA15']=pd.rolling_mean(index_data['closeIndex'],window=15)
index_data['MA20']=pd.rolling_mean(index_data['closeIndex'],window=20)
index_data['MA333']=pd.rolling_mean(pd.rolling_mean(index_data['closeIndex'],window=4),window=4)
index_data[['MA20','MA333']].plot(figsize=(12,6))
```

## 20日均线正斜率信息
**20日均线正斜率，斜率大于20%/250(年化20%为牛市，自己随意定义的，不满意可自己随意调整，250工作日)**


```python
index_data['MA20_DIFF']=index_data['MA20'].diff().fillna(method='bfill')
index_data['style']=(index_data['MA20_DIFF']*100/index_data['MA20']).apply(lambda x:1 if x>0.08 else(-1 if x<-0.0008 else 0))*300+3500
index_data[['style','MA20','closeIndex']].plot(figsize=(18,6))
```

**毛刺太多， 猴市偏少，大部分都是牛熊，不大符合实际**

## 平滑处理
**信号必须持续N天才算有效**

```python
#平滑去噪
index_data['MA20_DIFF']=index_data['MA20'].diff().fillna(method='bfill')
index_data['style']=(index_data['MA20_DIFF']*100/index_data['MA20']).apply(lambda x:1 if x>0.08 else(-1 if x<-0.08 else 0))
#信号必须持续n天才算有效
tmp=pd.Series(0,index=index_data['style'].index)
n_range=10
for i in range(n_range,len(index_data['style'])):
    if(index_data['style'][i-n_range:i].sum()==n_range):
        tmp[i]=1
    elif(index_data['style'][i-n_range:i].sum()==-n_range):
        tmp[i]=-1
    else:
        tmp[i]=0

index_data['style']=tmp*300+3500
index_data[['style','MA20','closeIndex']].plot(figsize=(18,6))
```

**信号变换太多，且明显滞后**

## 累计斜率试试
**其实等价于累计收益，分母可以看作不变，可近似理解为突破短期箱体**

```python
#平滑去噪
index_data['MA20_DIFF']=index_data['MA20'].diff().fillna(method='bfill')
index_data['style']=(index_data['MA20_DIFF']*100/index_data['MA20'])
#信号必须持续n天才算有效
tmp=pd.Series(0,index=index_data['style'].index)
n_range=10
for i in range(n_range,len(index_data['style'])):
    if(index_data['style'][i-n_range:i].sum()>n_range*0.08):
        tmp[i]=1
    elif(index_data['style'][i-n_range:i].sum()<-n_range*0.08):
        tmp[i]=-1

index_data['style']=tmp*300+3500
index_data[['style','MA20','closeIndex']].plot(figsize=(18,6))
```
