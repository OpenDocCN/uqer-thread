```python
import pandas as pd
import numpy as np
from pandas import DataFrame

#创建一个空的表格，并建立索引（名为date）
df = pd.DataFrame([], columns=['date','aa', 'bb', 'cc'],index=[])
df = df.set_index('date') #将列date设定为索引
print '1) 创建一个空的表格，并建立索引：'
print df
print len(df)
print
#然后，一行一行添加数据，并保持索引的有效性
new= pd.DataFrame({"aa":10,"bb":"b1","cc":20},index=["2017-01-03"])
df= df.append(new)
new= pd.DataFrame({"aa":20,"bb":"b2","cc":30},index=["2017-01-02"])
df= df.append(new)
new= pd.DataFrame({"aa":30,"bb":"b3","cc":40},index=["2017-01-03"])
df= df.append(new)
print '2) 连续添加了3行数据，注意有2行索引值相同：'
print df 
print
print '3）访问某列某行：先指定列名，后设定索引。这里的索引值有2行满足：'
print df['aa']['2017-01-03'] #“先列名，后索引”的访问方式
print
print '4) 取多列数据要用嵌套方式：'
print df[['cc','bb']] #取多列数据要用嵌套方式
print
print '5) 以索引得到行数据，这里得到2行：'
print df.loc['2017-01-03'] #以索引得到行数据
print
print '6) 在索引行的基础上，再指定列名：'
print df.loc['2017-01-03']['aa'] #以索引得到行数据，并限定某列
print
print '7) 采用嵌套方式，还可以指定多列：'
print df.loc['2017-01-03'][['aa','bb']] #以索引得到行数据，并限定多列
print
print '8) 索引的值竟然可以修改！'
df.index.values[0]= '2017-01-01' #修改第一行的索引值
print df.index.values #显示所有所有的值
print df
print

print '9）尝试创建一个类似context.history()的字典，其键的股票代码secID，值为一个表格（上述的df）'
my_dict={} #创建一个空字典
secID='000001.XSHE'
my_dict[secID] = df #在空字典中添加第一个元素
secID='600001.XSHG'
df2 = pd.DataFrame([[11,'ok','22']], columns=['aa', 'bb', 'cc'], index=['2017-01-05'])
df2.index.name= 'date'
my_dict[secID] = df2 #在字典中再添加一个元素
print my_dict

print 'ok!' #完成我的目标，分享给大家。
```
