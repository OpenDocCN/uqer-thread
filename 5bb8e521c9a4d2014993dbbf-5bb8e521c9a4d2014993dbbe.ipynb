```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import statsmodels.api as sm #这里利用statsmodel，还可以利用sklearn包
# 导入一元线性回归数据集
data1 = pd.read_excel(r'./WeChat/ModelTimes_Regression1.xlsx',sheetname=0)
# 导入多元线性回归数据集
data2 = pd.read_excel(r'./WeChat/ModelTimes_Regression1.xlsx',sheetname=1)
```

# 一元线性回归

```python
y = data1['Y'] #设置因变量
x = data1['X'] #设置自变量
X = sm.add_constant(x) #添加常数项列
```

```python
model = sm.OLS(y, X)
results = model.fit()
results.summary()
```

## 得到如下一元线性回归方程
$$y=2091.29+0.4375x$$

```python
#预测，2007年的人均可支配收入为95407.4亿元，预测2007年的消费支出为：43834.59372828
results.predict([1,95407.4])
```

```python
# 一元回归可视化
import seaborn as sns
sns.set(style="darkgrid")
plt.plot(x, results.predict(), label='pre_value')
plt.plot(x, y, 'o', label='true_value')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend(loc='best')
```

# 多元线性回归

```python
data2.columns = ['year', 'X', 'X1', 'GP', 'FP', 'Q', 'P0', 'P1']
y = np.log(data2['Q'])
X = np.log(data2[['X', 'P1', 'P0']])
X = sm.add_constant(X)
```

```python
model = sm.OLS(y, X)
results = model.fit()
results.summary()
```

## 得到如下多元线性回归方程
$$\ln{Q}=5.5320+0.5399\ln{X}-0.2580\ln{P_1}-0.2886\ln{P_0} $$

# 利用scikit-learn进行回归
## 以多元线性回归为例

```python
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score

model = linear_model.LinearRegression(fit_intercept=True)

# 取data2里面的X和P1，后面可视化最多为二维
X = np.array(np.log(data2[['X', 'P1']]))
y = np.array(np.log(data2['Q']))
model.fit(X,y)
#y的预测值
y_pred = model.predict(X)

# The coefficients
print('Coefficients:', model.coef_)
# The mean squared error
print("Mean squared error: %.2f"
      % mean_squared_error(y, y_pred))
# Explained variance score: 1 is perfect prediction
print('Variance score: %.2f' % r2_score(y, y_pred))
```

```python
#二元回归可视化平面
import seaborn as sns
sns.set(style="white")
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
from matplotlib import pyplot as plt
 
fig = plt.figure()
ax = Axes3D(fig)
ax.view_init(elev=15., azim=15)
x=np.log(data2['X'])
p1=np.log(data2['P1'])
X, P1 = np.meshgrid(x, p1)#网格的创建，这个是关键
Q=model.coef_[0]*X + model.coef_[1]*P1
plt.xlabel('X')
plt.ylabel('P1')
ax.plot_surface(X, P1, Q, rstride=1, cstride=1, cmap='rainbow')
plt.show()
```
