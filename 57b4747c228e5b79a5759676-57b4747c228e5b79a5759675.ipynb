```python
import numpy as np
import pandas as pd
import math
from datetime import datetime,timedelta
import matplotlib.pyplot as plot
import matplotlib.dates as dates

def _data_get(beginDate,endDate):
    bond = DataAPI.BondGet(secID=u"010107.XSHG",field=u"",pandas="1")
    maturityDate = bond[u'maturityDate'][0]
    coupon = bond[u'coupon'][0]
    print '起息日',bond['firstAccrDate']
    print '到期日', maturityDate
    print '利率', coupon

    data = DataAPI.MktBonddGet(secID=u"010107.XSHG",beginDate=beginDate,endDate=endDate,field=['tradeDate','closePrice'],pandas="1").set_index('tradeDate')
    tradeDate = data.index.tolist()
    closePrice = data['closePrice'].tolist()
    YTM = []
    for i in range(data.shape[0]):
        years = (datetime.strptime(maturityDate, '%Y-%m-%d') - datetime.strptime(tradeDate[i],'%Y-%m-%d')).days / 365.0
        ytm = (coupon*math.ceil(years) + 100  - closePrice[i])/closePrice[i] / years
        YTM.append(ytm)
    data['YTM'] = YTM 
    return data

#画图
def _plot(data):
    print data.tail(1)     # 最近1天数据
    ax = data.plot(figsize=(18,6),subplots=True,grid=True,rot=45,sharex=True,kind='line')



begin_date = "2014-04-01"
#end_date = "20141001"
end_date = datetime.today().strftime('%Y-%m-%d')  # 截止日期
_plot(_data_get(begin_date,end_date))
```
