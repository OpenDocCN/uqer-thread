## 从市盈率看股票是不是便宜

之前有在看[《投资中最简单的事》](https://item.jd.com/11566276.html)这边本书。其中，邱国鹭作者在说到如何买到便宜的股票时，有说过市盈率是一个股票是否便宜的重要指标。作为基本面量化的指标，我构建了一个沪深300的市盈率指数，研究了一下沪深300股票的市盈率，企图通过市盈率来看一下市场周期，观察目前情况下，股票是否真的便宜。

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
field = ['ticker', 'tradeDate', 'secShortName', 'closePrice', 'chgPct', 'PE', 'PE1']
```

```python
## 获取股票代码与权重
def get_300_ticker(begin, end):
    ##获取权重与数据
    df_300 = DataAPI.IdxCloseWeightGet(secID=u"",ticker=u"000300",beginDate= begin, endDate=end ,field=u"",pandas="1")
    ## 获取年最新数据
    df_300_weight = df_300.loc[df_300['consTickerSymbol'].drop_duplicates(keep = 'last').index][['consTickerSymbol', 'weight']]
    return {v[0]:v[1] for v in df_300_weight.values}

## 获取股票数据
def get_stocks_data(ticker,  begin, end, ticker_300 , filed = field):
    df_ori = DataAPI.MktEqudGet(secID=u"",ticker= ticker ,tradeDate=u"",beginDate=begin ,endDate= end,isOpen="1",
                   field= field,pandas="1")
    ## 股票数据*权重
    df_ticker = df_ori['PE1']*ticker_300[ticker]
    # df_ticker = df_ori['PE1']
    df_ticker.index = df_ori['tradeDate']
    df_ticker.name = ticker
    return df_ticker
```

```python
begin = '20170101'
end = '20180101'
## 从时间维度获取pe_index
def get_pe_index(begin, end):
    ticker_300 = get_300_ticker(begin, end)
    ticker_list = [get_stocks_data(ticker, begin, end, ticker_300 = ticker_300) for ticker in ticker_300.keys()]
    pe_index =  pd.concat(ticker_list, axis = 1).mean(axis = 1)
    return pe_index
```

```python
time_list = ['200%s0101'%(i) for i in range(5, 10)] + ['201%s0101'%(i) for i in range(9)]
```

```python
## 获取pe_index
index_list = []
for begin, end in zip(time_list[:-1], time_list[1:]):
    pe_index = get_pe_index(begin, end)
    index_list.append(pe_index)
    print(begin, end, 'has done')
```

```python
pe_index = pd.concat(index_list)
```

```python
pe_index.hist(bins = 30)
plt.show()
pe_index.plot()
plt.show()
```

## 初步结论
1. 从指数的分布来看，其分布多分布在10附近。严重的偏态分布。
2. 从指数的时间序列来看，其重要反弹大多集中在10附近。

记得书中说过一句，2013-2014年中旬蓝筹股集中在10左右非常久了。市场一度萎靡，这个时候市场有大把的便宜的股票。\
