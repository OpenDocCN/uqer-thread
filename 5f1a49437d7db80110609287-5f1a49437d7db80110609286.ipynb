```python
import numpy as np
import scipy as sp
import pandas as pd
```

## 1 获取多只股票多个交易日的指数变换的的DF 

```python
#第一步，获取多只股票多个交易日的指数变换的的DF
def get_allday_code():
    index_list = DataAPI.IndustryGet(industryVersion=u"SW",industryLevel=u"1",\
                                     isNew=u"1",industryID=u"",field=u"",pandas="1")['indexSymbol'].tolist()
    all_inf_list = []
    for code in index_list:
        one_inf_df = DataAPI.MktIdxdGet(indexID=u"",ticker=code,beginDate=u"20150101",\
                                        endDate=u"20200701",field=u"tradeDate,ticker,closeIndex",pandas="1")
        all_inf_list.append(one_inf_df)
    all_inf_df = pd.concat(all_inf_list,axis = 0)
    
    all_inf_df=all_inf_df.set_index(['tradeDate','ticker']).unstack('ticker')['closeIndex']
    
    return all_inf_df

all_inf_df = get_allday_code()    
all_inf_df.head()
```

## 2 获取每一个月第一个交易日的指数

```python
#第二步 获取每一个月第一个交易日的指数
from dateutil.parser import parse

all_inf_df['tradeDate'] = all_inf_df.index

all_inf_df.sort_values(by='tradeDate',inplace = True)

#df产生新的一列  parse 标准的时间标记方法
all_inf_df['tradeDate'] = all_inf_df['tradeDate'].map(lambda x:parse(x))
#df再产生新的一列，年月
all_inf_df['year_month'] = all_inf_df['tradeDate'].map(lambda x:(x.year,x.month))
#以年月为标准分组，并取第一行。获取每月第一个交易日信息的另一张方法。
month_inf_df = all_inf_df.groupby('year_month').head(1)

month_inf_df.head()
```

## 3 获取每一个交易日的涨跌幅排名

```python
#第三步，获取每一个交易日的涨跌幅排名
month_inf_df.drop('tradeDate',axis = 1,inplace = True)
month_inf_df.drop('year_month',axis = 1,inplace = True)

#pct_change和rank 均为在原来的DF上进行修改，使用时候注明axis = 0，1

month_inf_df = month_inf_df.pct_change(axis = 0)  
#pct_change计算增长率的更简单的方法

month_inf_df = month_inf_df[1:] #pct_change后第一行为NA

# 另一种计算肯德尔登记先关系数的方法rank，标记登记
month_inf_df = month_inf_df.rank(axis = 1)

month_inf_df.head()
```

## 4 定义获取相关系数函数，并且获取相关系数

```python
#第四步，定义获取相关系数函数，并且获取相关系数
import itertools
index_list = DataAPI.IndustryGet(industryVersion=u"SW",industryLevel=u"1",\
                                     isNew=u"1",industryID=u"",field=u"",pandas="1")['indexSymbol'].tolist()

#获取一届滞后相关系数的方法，corrcoef()里面必须为array的形式
def get_ind_corr(ind1,ind2,inf_df):
    y = np.array(inf_df[ind1][1:])
    x = np.array(inf_df[ind2][:-1])
    return np.corrcoef(y,x)[0][1]

ind_corr_dic ={}

for item in itertools.product(index_list,repeat = 2):  #itertools.product返回元组
    corr = get_ind_corr(item[0],item[1],month_inf_df)
    ind_corr_dic[item] = corr

ind_corr_dic
```

## 5 将结果转化为DF，并做图进行分析

```python
#第五步：将结果转化为DF

#字典转化为series  series再转化为DF
ind_corr_ser = pd.Series(ind_corr_dic)
#series再转化为DF
ind_corr_df = pd.DataFrame(ind_corr_ser)

ind_corr_df.reset_index(inplace= True)
ind_corr_df.columns  = ['ind1','ind2','corr']
ind_corr_df = ind_corr_df.sort_index(by = 'corr') #sort_index有时候可以打开多行占用一个index
ind_corr_df.head()
```

```python
#将结果转化名称
index_name_df = DataAPI.IndustryGet(industryVersion=u"SW",industryLevel=u"1",\
                                     isNew=u"1",industryID=u"",field=u"",\
                                    pandas="1")[['industryName','indexSymbol']]
#DF中的让未来字典中的关键字设定为idex
index_name_df.set_index('indexSymbol',drop = True,inplace = True)  
#生成字典
index_name_dict = dict(index_name_df['industryName'])
#map（）函数与字典的联合使用
ind_corr_df['ind1_name'] = ind_corr_df['ind1'].map(index_name_dict)

ind_corr_df['ind2_name'] = ind_corr_df['ind2'].map(index_name_dict)

#ind_corr_df = ind_corr_df[['ind1_name','ind2_name','corr']]

ind_corr_df
```
