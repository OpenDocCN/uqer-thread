# 关于50系列产品我能想到的

我的第三篇文章终于得到了超过20个评论，但是由于我带父母和小孩出门度假，所以一直到差不多30个评论才把这篇写好上传，实在不应该，所以今天至少2更。
这一篇应该属于大家说的水帖，但是如果你看进去了，应该还是很有用途的。
50etf没有停牌的风险，不像个股，如果停牌3个月再不走出个大的行情，这不是我们哭吗，白白耽误了我们的大好青春。
由于你懂得的原因，目前国内股指期货每个账户限制开仓的特别厉害，好像是10手每天，据说已经从几百万手的成交量，降低到几万手，而且为了限制投机者，当日平仓的手续费超贵，是普通手续费的几十倍，应该上千了，而新加坡的富时A50由于资金层面的问题，据说在9月份增加了一些，但是也有限，据说刘主席比较鼓励创新类的产品，什么时候能重新放开那，什么时候我们能直接投资海外那，让我们拭目以待。
股指期货曾经作为世界上最大的期货品种，让中国的期货公司繁荣了一段时间，而且好多期货公司都被上市的证券公司收购了，但由于股指期货的一蹶不振，期货公司好像现在有点不好受，多方位的积极寻找出路，比如大力开发期权客户，进行配资业务的尝试等等，反正如果你有什么特别的需求，别忘记去期货公司问问。
我最擅长的产品其实是期权，由于是上证50etf期权，主要由证券公司进行推广，但是好多期货公司也加入进来，目前全国只有10万个左右的户口，我个人觉得是一个机会。
如果你有50万以上的资金，可以先开个融资融券的账户，在通过考试开个期权的账户，再开个股指期货的账户，目前股票包括融资融券的佣金应该在万分之三以下，融资成本大概在百分之八左右，由于去年的几次降息据说有些证券公司能做到百分之六，个股期权大概8元每张，某些期货公司公司可以做到5元每张，股指期货当日为万分之二十五，隔日约万分之二。
目前的基准利率按百分之五计算，市面上的融资成本根据资金量，资金量大的大约为百分之十左右，小的民间借贷高达百分之二十，但是资源好像不很紧张，还是比较容易从市场找到钱。
