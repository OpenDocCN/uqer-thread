# 期权策略1120

《50ETF期权时间价值策略的实现》的验证

说明：策略所交易的标的为50ETF期权本身，而不是根据期权的成交数据来择时交易股票

而且由于没有期权策略的回测，所以所有的交易账户统计等只能自己写啦 策略思想参考了方正证券金融工程研究报告《20160506-方正证券-期权时间价值专题研究：一寸光阴一寸金，期权可买寸光阴》

策略组合：

卖出当月平值认购期权
卖出当月平值认沽期权
买入下月平值认购期权
买入下月平值认沽期权 （相当于2个日历差价合约组合，或者2个跨式合约组合)


换仓条件：ETF收盘价超过现持有的平值合约行权价格，换仓为最接近ETF收盘价的平值合约
回测时间:

时间区间从2015年2月9日到now

```python
import pandas as pd
dd=DataAPI.OptDpoGet(secID=u"",optID=u"",ticker=u"",varSecID=u"",varTicker=u"510050",varType=u"",beginDate=u"",endDate=u"",exerType=u"",contractType=u"",field=u"",pandas="1")

data = dd[["tradeDate","optID","secShortName","ticker","strikePrice",\
           "listDate","expDate","preClosePrice","contractType",\
           "varpreClosePrice","marginUnit"]]
data["diff"]=abs(data["strikePrice"]-data["varpreClosePrice"]).round(3) 
del  dd;

#trading day，获得每个交易日的日期
TradingDaysSeries=data["tradeDate"]
TradingDaysList=[]
for i in TradingDaysSeries:
    TradingDaysList.append(i)
TradingDaysList=list(set(TradingDaysList))
TradingDaysList=sorted(TradingDaysList)
del TradingDaysSeries

# 50ETFPrice，获得每个交易日的ETF价格
ETFPrice=[]
for i in TradingDaysList:
    Trimming=data[data.tradeDate==i]
    ETFPrice.append(Trimming.iloc[0]["varpreClosePrice"])


ETFPrice2=zip(TradingDaysList,ETFPrice)# list(Tuple(time,price))
ETFPriceList = list( (name,value) for name,value in ETFPrice2)
ETFPriceDict = dict( (name,value) for name,value in ETFPrice2)
del ETFPrice2,ETFPrice,Trimming

##  求每天的平值期权，近月，远月，call，put,放入Holding中（dataframe）
Holding=pd.DataFrame(columns=[["tradeDate","optID","secShortName","ticker","strikePrice",\
           "listDate","expDate","preClosePrice","contractType",\
           "varpreClosePrice","marginUnit","diff"]]);

for i in TradingDaysList:
    #选出当天所有的call
    Trimming=data[data["tradeDate"]==i]
    ExpDateSeries=Trimming["expDate"]
    ExpDateList = []
    for j in ExpDateSeries:
        ExpDateList.append(j)
    ExpDateList=list(set(ExpDateList))
    ExpDateList=sorted(ExpDateList)
    CurrentMonth=ExpDateList[0]
    NextMonth=ExpDateList[1];
    #选出Trimming中ExpDate==CurrentMonth的call,并挑出平值期权
    TrimmingTemp1=Trimming[Trimming["expDate"]==CurrentMonth][Trimming["contractType"]=="CO"]
    TrimmingTemp1Sort=TrimmingTemp1.sort(columns=["diff"])
    CrtMonthCall=TrimmingTemp1Sort.iloc[0:1]
    Holding=Holding.append(CrtMonthCall,ignore_index=True)
    #选出Trimming中ExpDate==CurrentMonth的put,并挑出平值期权
    TrimmingTemp1=Trimming[Trimming["expDate"]==CurrentMonth][Trimming["contractType"]=="PO"]
    TrimmingTemp1Sort=TrimmingTemp1.sort(columns=["diff"])
    CrtMonthPut=TrimmingTemp1Sort.iloc[0:1]
    Holding=Holding.append(CrtMonthPut,ignore_index=True)
    #选出Trimming中ExpDate==NextMonth的call,并挑出平值期权
    TrimmingTemp1=Trimming[Trimming["expDate"]==NextMonth][Trimming["contractType"]=="CO"]
    TrimmingTemp1Sort=TrimmingTemp1.sort(columns=["diff"])
    NextMonthCall=TrimmingTemp1Sort.iloc[0:1]
    Holding=Holding.append(NextMonthCall,ignore_index=True)
    #选出Trimming中ExpDate==NextMonth的put,并挑出平值期权
    TrimmingTemp1=Trimming[Trimming["expDate"]==NextMonth][Trimming["contractType"]=="PO"]
    TrimmingTemp1Sort=TrimmingTemp1.sort(columns=["diff"])
    NextMonthPut=TrimmingTemp1Sort.iloc[0:1]
    Holding=Holding.append(NextMonthPut,ignore_index=True)
    # print Holding

# Holding.to_excel("Holding.xlsx");
## 处理平值期权到期日
#以上处理完的Holding有这样的问题：tradeDate=20150325时，对于的近月call行权日为0325，
#对于这种情况，不开仓

Holding.head(20)
```

```python
# 收益统计
ProfitList=[0];
# 手续费统计，手续费按最低的每张2.3元算，这里买多，卖空，平多，平空都算手续费了，实际上卖空是不要手续费的
# 这里是为了让手续费更为苛刻。
CommissionList=[2.3*4];
for i in TradingDaysList:
    if i==TradingDaysList[0]:
        PreviousI=i
        continue
    #计算盈亏情况
    Profit=0;
    Trimming=Holding[Holding["tradeDate"]==PreviousI]
    if Trimming["tradeDate"].values[0]!=Trimming["expDate"].values[0]:
        ## 判断是否换仓，换仓的话就计算手续费
        if Trimming.iloc[0:1]["ticker"].values[0]!=Holding[Holding["tradeDate"]==i].iloc[0:1]["ticker"].values[0]:
            # print ("change ticker")
            CommissionList.append(2.3*8)
        else:
            CommissionList.append(0);
        #近月call收益
        YesterdayTicker=Holding[Holding["tradeDate"]==PreviousI].iloc[0:1]["ticker"].values[0]
        CrtCallTodayPrice=data[data["tradeDate"]==i][data["ticker"]==YesterdayTicker]["preClosePrice"].values[0]
        CrtCallYesterdayPrice=Holding[Holding["tradeDate"]==PreviousI].iloc[0:1]["preClosePrice"].values[0]
        Profit=Profit+CrtCallYesterdayPrice-CrtCallTodayPrice;
        # 近月put收益
        YesterdayTicker=Holding[Holding["tradeDate"]==PreviousI].iloc[1:2]["ticker"].values[0]
        CrtPutTodayPrice=data[data["tradeDate"]==i][data["ticker"]==YesterdayTicker]["preClosePrice"].values[0]
        CrtPutYesterdayPrice=Holding[Holding["tradeDate"]==PreviousI].iloc[1:2]["preClosePrice"].values[0]
        Profit=Profit+CrtPutYesterdayPrice-CrtPutTodayPrice;
        #远月call收益
        YesterdayTicker=Holding[Holding["tradeDate"]==PreviousI].iloc[2:3]["ticker"].values[0]
        NextCallTodayPrice=data[data["tradeDate"]==i][data["ticker"]==YesterdayTicker]["preClosePrice"].values[0]
        NextCallYesterdayPrice=Holding[Holding["tradeDate"]==PreviousI].iloc[2:3]["preClosePrice"].values[0]
        Profit=Profit+NextCallTodayPrice-NextCallYesterdayPrice;
        #远月Put收益
        YesterdayTicker=Holding[Holding["tradeDate"]==PreviousI].iloc[3:4]["ticker"].values[0]
        NextPutTodayPrice=data[data["tradeDate"]==i][data["ticker"]==YesterdayTicker]["preClosePrice"].values[0]
        NextPutYesterdayPrice=Holding[Holding["tradeDate"]==PreviousI].iloc[3:4]["preClosePrice"].values[0]
        Profit=Profit+NextPutTodayPrice-NextPutYesterdayPrice;
        #当日收益情况
        PreviousI=i;
        ProfitList.append(Profit)
    else:
        PreviousI=i;
        ProfitList.append(0);

ProfitSum=sum(ProfitList)
ProfitSumList=[]
CommissionSumList=[]
for i in range(len(ProfitList)):
    ProfitSumList.append(sum(ProfitList[0:i])*10000)
    CommissionSumList.append(sum(CommissionList[0:i]))

import matplotlib.pyplot as plt
plt.plot(ProfitSumList)
plt.figure()
plt.plot(CommissionSumList)
```
