# 春节行情统计

明天（20160204）是羊年最后一个交易日，A股素来有春节红包行情之称，春节前后的行情值得期待。

下面统计一下A股以及标普，恒指在春节前1个交易日，以及春节后5，10，20个交易日的涨跌幅情况。

羊年跌宕起伏，uqer一路相伴，祝uqers们猴年发发发！

```python
from CAL.PyCAL import Calendar
import datetime as dt
import pandas as pd

cal = Calendar('China.SSE')

start_date = '20050101'
columns = ['000001','000300', 'SPX','HSI']

# 获取上证，HS300，标普，恒指在05年以后的收盘价。
data = DataAPI.MktIdxdGet(ticker=columns, beginDate=start_date, field='ticker,tradeDate,closeIndex').pivot(index='tradeDate',columns='ticker',values='closeIndex')
# data.fillna(method='pad', inplace=True)
data.head()
```

```python
# 获取上交所交易日历。
df = DataAPI.TradeCalGet('XSHG', field='isOpen,calendarDate,prevTradeDate')
df.set_index('calendarDate',inplace=True)
df = df.merge(data,left_index=True, right_index=True)

# 去除非上交所交易的交易日
df = df[df.isOpen==1]

# 计算两个交易日的天数差，如果大于等于7天，可能就是春节了^^
df['calendarDate'] = pd.to_datetime(df.index)
df['lastIndex'] = (df.calendarDate - df.calendarDate.shift())
df['month'] = df.calendarDate.apply(lambda x:x.month)
df['year'] = df.calendarDate.apply(lambda x:x.year)

#再去除掉国庆节，劳动节（好怀念以前的七天劳动节。。）
df[(df.lastIndex>=pd.Timedelta(7,unit='d')) & (df.month!=10)& (df.month!=5)]
```

```python
# 每年只有1，2月份出现一次7天长假，我认为这就是春节了，有意见的同学可以自己对照老黄历。

# 下面开始计算不同周期的涨跌幅了
dfs = {}
periods = [-1, 1, 5, 10, 20]

for period in periods:
    tmp_df = df[columns+['calendarDate', 'prevTradeDate', 'lastIndex', 'month']]
    for col in columns:
        if period < 0:
            tmp_df['%s %sdays' %(col,period)] = ((df[col].shift(1)/df[col].shift(-period+1)-1)*100).apply(lambda x: (round(x,2)))
            tmp_df['%s %s' %(col,period)] = df[col].shift(1)
            tmp_df['%s %s2' %(col,period)] = df[col].shift(-period+1)
        else:
            tmp_df['%s %sdays' %(col,period)] = ((df[col].shift(1-period)/df[col].shift(1)-1)*100).apply(lambda x: (round(x,2)))
            tmp_df['%s %s' %(col,period)] = df[col].shift(1-period)
        
    tmp_df = tmp_df[(tmp_df.lastIndex>=pd.Timedelta(7,unit='d')) & (tmp_df.month!=10)& (tmp_df.month!=5)]
    dfs[period] = (tmp_df)
    
f = lambda x: '\n'.join(['    指数:%s, 涨幅:%s%%' %(i, x[i]) for i in x.index])
for k in periods:
    v = dfs[k]
    if k<0:
        print '春节前%s天：' %abs(k)
    else:
        print '春节后%s天：' %k
    d = v[['%s %sdays' %(col,k) for col in columns]].describe()
    print '最大涨幅：\n%s\n最小涨幅：\n%s\n平均涨幅：\n%s\n涨幅中位数：\n%s' %(f(d.iloc[-1]), f(d.min()), f(d.mean()), f(d.iloc[5]))
    print v.to_html()
```
