```python
def plot_k(quotes, sec):
    __color_balck__= '#0F0F0F'
    __color_green__= '#00FFFF'
    __color_yellow__ = '#EE9A00'
    __color_purple__ = '#9900CC'
    __linewidth__ = 2

    fig = plt.figure(figsize=(11,6))
    fig.set_tight_layout(True)
    
    # font = matplotlib.font_manager.FontProperties(fname=r"C:\Windows\Fonts\STFANGSO.TTF")  #fname指定字体文件  选简体显示中文
    
    ax1 = fig.add_axes([0, 1, 1, 1])
    ax1.set_title(u'K线图', fontproperties=font, fontsize=20)
    ax2 = fig.add_axes([0, 0.35, 1, 0.5], axis_bgcolor='w')
    ax1.set_axisbelow(True)
    ax2.set_axisbelow(True)
    
    ax1.grid(True, axis='y')
    ax2.grid(True, axis='y')
    ax1.set_xlim(-1, len(quotes)+1)
    ax2.set_xlim(-1, len(quotes)+1)

    for i in range(len(quotes)):
        close_price = quotes.ix[i, 'closePrice']
        open_price = quotes.ix[i, 'openPrice']
        high_price = quotes.ix[i, 'highestPrice']
        low_price = quotes.ix[i, 'lowestPrice']
        vol = quotes.ix[i, 'turnoverVol']
        trade_date = quotes.ix[i, 'tradeDate']
        if close_price > open_price:
            ax1.add_patch(patches.Rectangle((i-0.2, open_price), 0.4, close_price-open_price, fill=False, color='r'))
            ax1.plot([i, i], [low_price, open_price], 'r')
            ax1.plot([i, i], [close_price, high_price], 'r')
            ax2.add_patch(patches.Rectangle((i-0.2, 0), 0.4, vol, fill=False, color='r'))
        else:
            ax1.add_patch(patches.Rectangle((i-0.2, open_price), 0.4, close_price-open_price, color='g'))
            ax1.plot([i, i], [low_price, high_price], color='g')
            ax2.add_patch(patches.Rectangle((i-0.2, 0), 0.4, vol, color='g'))
    # ax1.set_title(sec,  fontsize=15, loc='left', color='r')
    # ax2.set_title(u'成交量',  fontsize=15, loc='left', color='r')

    ax1.set_xticks(range(0,len(quotes), 15))
    ax2.set_xticks(range(0,len(quotes), 15)) 
    s1 = ax1.set_xticklabels([mdates.num2date(quotes.ix[index, 'tradeDate']).strftime('%Y-%m-%d') for index in ax1.get_xticks()])
    s1 = ax2.set_xticklabels([mdates.num2date(quotes.ix[index, 'tradeDate']).strftime('%Y-%m-%d') for index in ax2.get_xticks()])
         
    ma20 = pd.rolling_mean(np.array(quotes['closePrice'], dtype=float), window=20, min_periods=0)
    ma60 = pd.rolling_mean(np.array(quotes['closePrice'], dtype=float), window=60, min_periods=0)
    ma250 = pd.rolling_mean(np.array(quotes['closePrice'], dtype=float), window=250, min_periods=0)

    ax1.plot(ma20, color='b', linewidth=__linewidth__)
    ax1.plot(ma60, color=__color_yellow__, linewidth=__linewidth__)
    ax1.plot(ma250, color=__color_purple__, linewidth=__linewidth__)
    
    ax1.annotate('MA20-', xy=(len(quotes)-30, ax1.get_yticks()[-1]), color='b', fontsize=15)
    ax1.annotate('MA60-', xy=(len(quotes)-19, ax1.get_yticks()[-1]), color=__color_yellow__, fontsize=15)
    ax1.annotate('MA250-', xy=(len(quotes)-8, ax1.get_yticks()[-1]), color=__color_purple__, fontsize=15)
    vol5 = pd.rolling_mean(np.array(quotes['turnoverVol'], dtype=float), window=5, min_periods=0)
    vol10 = pd.rolling_mean(np.array(quotes['turnoverVol'], dtype=float), window=10, min_periods=0)
    ax2.plot(vol5, color='b', linewidth=__linewidth__)
    ax2.plot(vol10, color=__color_yellow__, linewidth=__linewidth__)

    return fig
```

```python
def get_historical_closeIndex(ticker, start, end):
    # get the data for the tickers.  This will be a panel
    p = DataAPI.MktIdxdGet(indexID=ticker,beginDate=start,endDate=end,field=u"tradeDate,indexID,closeIndex",pandas="1")
    p['tradeDate'] = pd.to_datetime(p['tradeDate']) #因为DataAPI所有日期都是以字符串形式存储的，因此此处需要转换成日期时间型,这里很重要！
    # rename the columns
    p.rename(columns={'tradeDate': 'Date', 
                      'closeIndex': 'Close'}, inplace=True)
    # pivot each ticker to a column
    pivoted = p.pivot(index='Date', columns='indexID')
    # and drop the one level on the columns
    pivoted.columns = pivoted.columns.droplevel(0)
    return pivoted
```

```python
# 技术因子
# Bollinger Bands
def get_bbands(df, ndays):
    db = df.copy()
    dm = df.rolling(ndays).mean()
    ds = df.rolling(ndays).std()
    db['upperBB'] = dm + 2 * ds
    db['lowerBB'] = dm - 2 * ds
    return db

# Simple Moving Average
def get_SMA(df, ndays):
    dm = df.copy()
    dm.rolling(ndays).mean()
    return dm

# Expontential Moving Average
def get_EMA(df, ndays):
    dm = df.ewm( span = ndays, min_periods = ndays - 1).mean()
    return dm

# Rate of Change
def get_ROC(df, ndays):
    dn = df.diff(ndays)
    dd = df.shift(ndays)
    dr = dn/dd
    return dr

# dates = pd.date_range('2016-01-01','2016-12-31',freq='B')
# symbols = ['aapl']
# df = stocks_data(symbols, dates)
df = normalized_price
dm = get_SMA(df, 10)
dm.plot(figsize=(16, 9))
plt.title('Simple Moving Average')
plt.show()
dm = get_EMA(df, 10)
dm.plot(figsize=(16, 9))
plt.title('Expontential Moving Average')
plt.show()

dr = get_ROC(df, 1)
dr.plot(figsize=(16, 9))
plt.title('Rate of Change')
plt.show()
```
