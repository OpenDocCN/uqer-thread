# 2.策略的理论基础
跨期套利是指投资者以赚取期货合约价差为目的，在同一期货品种的不同月份合约上建立数量相等、方向相反的交易头寸，最后以对冲或交割方式结束交易、获得收益的方式。最简单的跨期套利就是买入近期的期货品种，卖出远期的期货品种。因此跨期套利所关注的是合约之间的价差，能够一定程度上避免单个期货合约价格剧烈变动所带来的风险。  
对于同一个投资标的，市场上一般有若干张不同到期月份的期货合约。以上证50指数为例，同一时刻，市场上存在当月，下月，当季，下季四个股指期货。这些合约均有同一现货标的，受相同因素的影响，所以合约价格表现出一定的相关性。在市场预期一致的情况下，它们之间的价差应该是稳定的，一旦发生了偏离，那么就出现了套利的机会。  
相比期现套利来说，跨期套利并不是无风险套利。期现套利的价差会在合约到期时收敛，但是跨期套利的价差可能会随着股指的上升而上升，实际中也与投资者对中期或短期股指的预期相关，如果价差在一个新的水平上达到平衡，那么我们的策略便面临着风险。


我们拿2015年1月4日到2015年12月31日的上证50股指期货当月合约和下月合约的分钟数据，来对价差进行单位根检验。函数get\_current\_contracts可以根据当前时间和标的获取标的对应的当前四个合约的名称。

```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
def third_friday(current_time):
    first_day = pd.datetime(current_time.year, current_time.month, 1)
    if first_day.isocalendar()[2] == 5:
        time = 1
    else:
        time = 0
    for i in range(31):
        next_day = first_day + timedelta(i+1)
        if next_day.isocalendar()[2] == 5:
            time += 1
        if time == 3:
            break
    return next_day

def get_current_contracts(current_time, types):
    if current_time > third_friday(current_time):
        current_month = pd.datetime(current_time.year+int((current_time.month+1)>12), current_time.month+1, 1)
    else:
        current_month = pd.datetime(current_time.year, current_time.month, 1)
    next_month = pd.datetime(current_month.year+int((current_month.month+1)>12), 
                             (current_month.month+1)%12+12*((current_month.month+1)/12-int((current_month.month+1)>12)), 1)
    current_season = pd.datetime(next_month.year+int(next_month.month>=12), 
                                 ((next_month.month+3)/3*3)%12+12*int((((next_month.month+3)/3*3)%12==0)), 1)
    next_season = pd.datetime(current_season.year+int(current_season.month>=12),
                              ((current_season.month+3)/3*3)%12+12*int((((current_season.month+3)/3*3)%12==0)),1)
    current_month_contracts = types+datetime.strftime(current_month, '%y')+datetime.strftime(current_month, '%m')
    next_month_contracts = types+datetime.strftime(next_month, '%y')+datetime.strftime(next_month, '%m')
    current_season_contracts = types+datetime.strftime(current_season, '%y')+datetime.strftime(current_season, '%m')
    next_season_contracts = types+datetime.strftime(next_season, '%y')+datetime.strftime(next_season, '%m')
    return [current_month_contracts, next_month_contracts, current_season_contracts, next_season_contracts]
```

如下图所示，我们选取了从2016年1月1日至2016年6月1日的当月合约与下月合约分钟价差数据，价差大概稳定在25个点。再对其进行adf-fuller单位根检验， p值为0.0013，拒绝原假设，即认为序列平稳。价差的稳定保证了价差出现异常后会回复，其围绕均衡价差的波动带来了套利机会，价差从偏离到回复的收益是可以把握的。

```python
import matplotlib.pyplot as plt
a = np.array(current_month)
b = np.array(next_month)
fig = plt.figure(figsize=(10,4))
fig.set_tight_layout(True)
plt.plot(b-a, '-b')
plt.plot([-20]*len(b-a), 'r')
```

```python
from statsmodels.tsa.stattools import adfuller
test = adfuller(b-a)
result = pd.Series(test[0:4], index=['Test Statistic','p-value','Lags Used','Number of Observations Used'])
for key,value in test[4].items():
    result['Confidence level (%s)'%key] = value
pd.DataFrame(result, columns=['diff'])
```

因为当前上证50 股指期货的成交较为活跃仅有当月合约和下月合约，需要使用三张甚至更多合约的蝶式或者鹰式策略并不适用，因此这里和研报一样选取牛熊市跨期套利作为基本策略。  

```python
from CAL.PyCAL import font
font.set_size(10)
import DataAPI
current_month_vol = DataAPI.MktFutdGet(tradeDate=u"20160513", ticker=[u"IH1605"], field=[u"turnoverVol"], pandas="1").iat[0, 0]
next_month_vol = DataAPI.MktFutdGet(tradeDate=u"20160513", ticker=[u"IH1606"], field=[u"turnoverVol"], pandas="1").iat[0, 0]
current_season_vol = DataAPI.MktFutdGet(tradeDate=u"20160513", ticker=[u"IH1609"], field=[u"turnoverVol"], pandas="1").iat[0, 0]
next_season_vol = DataAPI.MktFutdGet(tradeDate=u"20160513", ticker=[u"IH1612"], field=[u"turnoverVol"], pandas="1").iat[0, 0]
plt.bar([0, 1, 2, 3], [current_month_vol, next_month_vol, current_season_vol, next_season_vol], align="center", width=0.5) 
plt.xticks([0, 1, 2, 3], [u'当月合约', u'下月合约', u'当季合约', u'下季合约'], fontproperties=font)
```

牛市跨期套利是指投资者判断市场向上，价差随之上升时，从而采用做多远月合约、做空近月合约的组合进行套利的策略。而熊市套利是指投资者判断市场向下，价差随之下降时，从而采用做多近月合约、做空远月合约的组合进行套利的策略。

```python
fig = plt.figure(figsize=(5,5))
fig.set_tight_layout(True)
ax1 = fig.add_subplot(211)
ax2 = fig.add_subplot(212)
x = np.array([1,2,3,4,5])
y1 = 2*x
y2 = 3*x+1
ax1.plot(x, y1, label='F1')
ax1.plot(x, y2, label='F2')
ax1.legend(loc='left')
ax1.set_xlim(0, 6)
ax1.set_ylim(0, 18)
ax1.set_title(u'牛市套利示意图', fontproperties=font, fontsize=16)

y4 = 0.5*x+10
y3 = 1.5*x+4
ax2.plot(x, y3, label='F1')
ax2.plot(x, y4, label='F2')
ax2.legend(loc='left')
ax2.set_xlim(0, 6)
ax2.set_ylim(0, 18)
ax2.set_title(u'熊市套利示意图', fontproperties=font, fontsize=16)
```

我们已经证明价差是平稳的，所以基于牛熊市跨期套利对价差的判断就来自于价差偏离均衡值后会回复的特性。这里以2倍标准差做建仓标准为例, 具体的操作步骤如下:  
- 牛市跨期套利:当价差从下往上穿越负2倍标准差时建仓，可以预期价差要恢复至均衡值，做空当月合约，做多下月合约；当价差从下往上穿越均衡值时平仓。  
- 熊市跨期套利:当价差从上往下穿越正2倍标准差时建仓，可以预期价差要恢复至均衡值，做多当月合约，做空下月合约；当价差从上往下穿越均衡值时平仓。  

**关于均衡价差的计算**

- 1.模拟交割法：
通过模拟交割过程各项费用，来计算套利过程中的各项成本，从而得出均衡价差  
优点：容易计算，价差变化下；缺点：实际中价差很少恢复到这一水平

- 2.历史滞后均值法：
通过历史的价差，选择时间窗口长度，得到均衡价差  
优点：容易计算，价差动态；缺点：不能保证样本外价差恢复到这一水平

这里，我们选择了历史滞后均值法来计算均衡价差，一方面它计算简单，另一方面可以随着新增的数据动态的变化。初始资金设置为100w，以牛市跨期套利为例。

# 5.一些补充
- 是否需要添加止损策略  
这里选取均衡价差是一个动态的历史均值，可以较好的缓冲价差的变化，不强求价差稳定在一个数值上，不需要使用止损策略。

- 冲击成本的影响  
这里我们没有考虑冲击成本的影响，但是实际操作过程中，股指期货每分钟价格的变化有可能会达到50多，因此冲击成本会成为影响股指期货跨期套利的重要因素。  

总的来说，均值策略是统计方法中较为容易操作且收益不错的方法。
