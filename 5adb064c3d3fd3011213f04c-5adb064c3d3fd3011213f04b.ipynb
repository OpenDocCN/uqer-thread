# 1.SMA双均线简单量化思路及优化

SMA（simple moving average）,简单移动平均线，是最简单而普遍的移动平均线。
平均线是指算术平均数，计算方法为一组数字相加，除以该组数据的组成个数，其中每一给定时限在计算平均值时的权重均相等 。
以5天移动平均线为例，公式如下：

MA=(C1+C2+C3+C4+C5)/5
一般公式 :MA=(C1+C2+C3+C4+C5+....+Cn)/n

C: 第一日收盘价
n: 移动平均数周期

当短期均线上穿长期均线时，视为做多信号，相反则视为做空信号。
均线策略在趋势型市场中效果较好，而在震荡型市场中效果欠佳。
类似的指标如EMA，MACD等都可使用本文的思路进行研究。

本文的研究思路按照如下的五个步骤进行：
一.数据准备		
								 二.策略思想
                                 三.计算策略收益
                                 四.基础回测指标构建
                                 五.优化思路
                              

一.数据准备
	以沪深300为投资标的，通过优矿的DataAPI.MktIdxdGet接口获取数据

```python
import numpy as np
import pandas as pd

hs300_data = DataAPI.MktIdxdGet(indexID=u"",ticker=u"000300",tradeDate=u"",beginDate=u"20100101",endDate=u"20180101",exchangeCD=u"XSHE,XSHG",field=u"",pandas="1")
hs300_data.head()

# hs300_data['closeIndex'].plot(figsize = (10, 6))            
#画出HS300的指数图
```

二.策略思想
	当短期均线上穿长期均线时，视为做多信号，仓位表示为1；反之视为做空信号，仓位表示为-1
    在这里我们使用10日线和60日线分别作为短期线和长期线

```python
import talib as ta                                              
#导入技术分析库talib

short = 10                                                       
long = 60
#将日期数设为全局变量，方便调参

hs300_data['sma_short'] = ta.SMA(np.asarray(hs300_data['closeIndex']), short)           
# hs300_data['sma__short'] = hs300_data['closeIndex'].rolling(short).mean()     
#rolling的方法也可以用来计算均线，结果相同

hs300_data['sma_long'] = ta.SMA(np.asarray(hs300_data['closeIndex']), long) 
# hs300_data['sma__long'] = hs300_data['closeIndex'].rolling(long).mean()

hs300_data['position'] = np.where(hs300_data['sma_short'] > hs300_data['sma_long'], 1, -1)   
#where语句，当判断结果为True时，赋值为1，否则赋值为-1 
#新建一列'position'表位为仓位

hs300_data.dropna(inplace = True)                                                             
#当未满足10天或60天时，无法求出均线数据，删除掉这些行

hs300_data.tail()
```

三.计算策略收益
	pct_change语句可计算每日收益
    将每日收益乘以仓位，即可得到策略收益
    通过对'position'进行shift操作，可避免未来函数
    收益先累计求和，然后作为e的指数，即为累计收益率

```python
hs300_data['return'] = hs300_data['closeIndex'].pct_change()

# hs300_data['return'].hist(bins = 40)                                                         
#可画出每日收益率的柱状分布图
                             
hs300_data['strategy'] = hs300_data['position'].shift(1) * hs300_data['return']

hs300_data[['return', 'strategy']].cumsum().apply(np.exp).plot(figsize = (8, 6))               
#可画出累计收益率图
```

四.基础回测指标构建
	这里只构建了最基础的三个回测指标：年化收益率 波动率 最大回撤

```python
return_year = hs300_data[['return', 'strategy']].mean() * 252                         
#年化收益率

volatility = hs300_data[['return', 'strategy']].std() * 252 ** 0.5                    
#波动率

hs300_data['cumreturn'] = hs300_data['strategy'].cumsum().apply(np.exp)         
hs300_data['cummax'] = hs300_data['cumreturn'].cummax()
drawdown = (hs300_data['cummax'] - hs300_data['cumreturn']).max()                     
#最大回撤，即为策略到达的历史最高收益率与当前累计收益率的差值的最大值

print pd.concat((pd.DataFrame(return_year, columns=['return_year']), pd.DataFrame(volatility, columns=['volatility'])), axis = 1)    
#将benchmark即HS300和策略的年化收益率、波动率整合在一张DataFrame中
                                                                                                                                     
print 'drawdown is ' + str(drawdown)
```

五.优化思路
	之前的思路是：只要短期均线和长期均线有差异，我们就开仓。
    由于均线策略是趋势型量化策略，当市场处于震荡时期时，趋势未形成，频繁地开仓不仅有很高的交易成本，而且会使得策略无效。
    简单的优化思路是增加一个过滤器，当短期、长期均线差异较大时，我们认为趋势较为明显，此时再开多仓或空仓。无此信号时，平仓。

```python
hs300_data['diff'] = hs300_data['sma_short'] - hs300_data['sma_long']                                    
#计算每日短期、长期均线的差

sd = 20                                                                                                  
#设定阈值，此处为固定值，也可设计为指数百分比的移动阈值

hs300_data['position_opt'] = np.where(hs300_data['diff'] > sd, 1, 0)                                    
hs300_data['position_opt'] = np.where(hs300_data['diff'] < -sd, -1, hs300_data['position_opt'])
#where语句判断仓位，0表示空仓，opt为optimize的简称

hs300_data['strategy_opt'] = hs300_data['position_opt'].shift(1) * hs300_data['return']

hs300_data[['return', 'strategy_opt']].cumsum().apply(np.exp).plot(figsize = (8, 6))
```

```python
return_year = hs300_data[['return', 'strategy', 'strategy_opt']].mean() * 252                           

volatility = hs300_data[['return', 'strategy', 'strategy_opt']].std() * 252 ** 0.5                      

hs300_data['cumreturn_opt'] = hs300_data['strategy_opt'].cumsum().apply(np.exp)         
hs300_data['cummax_opt'] = hs300_data['cumreturn_opt'].cummax()
drawdown_opt = (hs300_data['cummax_opt'] - hs300_data['cumreturn_opt']).max()                       

print pd.concat((pd.DataFrame(return_year, columns=['return_year']), pd.DataFrame(volatility, columns=['volatility'])), axis = 1)    

print 'drawdown is ' + str(drawdown)

print 'drawdown after optimization is ' + str(drawdown_opt)
```

可以看出，经过优化后的策略，在年化收益率和波动率方面都有了提升，而且交易次数肯定也有减少。不幸的是，最大回撤反而变大了，这是我们不想看到的。
