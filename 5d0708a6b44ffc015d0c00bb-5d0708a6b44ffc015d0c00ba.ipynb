```python
#http://www.quantatrisk.com/2014/10/23/garch11-model-in-python/
#这里的garch是争对价格的波动率即日内涨跌幅建模
from arch import arch_model
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt

target='510050.XSHG'
start='20170103'
final='20190612'
data=DataAPI.MktFunddGet(secID=target,ticker=u"",tradeDate=u"",beginDate=start,endDate=final,field=u"",pandas="1")
def tolog(arr):
    arr=np.array(arr)
    tmp=sp.log(arr[1:]/arr[0:-1])
    return tmp
log_return=tolog(data.closePrice)

train=log_return[:560]

garch=arch_model(train,p=1,o=0,q=1,vol='Garch',dist='Normal')
result = garch.fit(update_freq=5)
#print(result.summary())
#volatility=(result.conditional_volatility)
#print(len(volatility))#560
C, A, B = result.params[1:] # 保存 GARCH 系数
def arch_p(C, A, B, returns, p):
    #基于GARCH(1,1)模型模拟的系数，返回  ARCH(p) 模型中的 sigma^2  
    squared_returns = returns**2
    lagged_squared_returns = squared_returns[-p:] #
    lagged_squared_returns = lagged_squared_returns[::-1] # 顺序颠倒
    a = C/(1-B)
    a_i = [A*B**(i-1) for i in range(1,p+1)]

    ans = a + np.dot(a_i,lagged_squared_returns)
    return ans
def garch_forecast(C, A, B, returns, dist='Normal'):
   #GARCH模型 L1 预测 
    #C GARCH 模型的常数项
    #A GARCH 模型的 Y_n^2 系数
    #B GARCH 模型的 sigma_n^2 系数    
    n = len(returns)
    p = int(n/2)
    #GARCH 模型的 sigma_n^2 系数
    arch_p_term = arch_p(C, A, B, returns, p)
    #计算残差项平方的  median 
    if dist.lower()=='normal':
        Zn2 = 0.45
    elif dist.lower()=='t':
        Zn2 = 0.53
    else:
        print("无效分布输入, 只能选 'normal' 或 't' ")
        return None  
    forecast = arch_p_term * Zn2
    return forecast
out=garch_forecast(C,A,B,train)
print(round(out,4))
print(data['CHGPct'][561])
```
