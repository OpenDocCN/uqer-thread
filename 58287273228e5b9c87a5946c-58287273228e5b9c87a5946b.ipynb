再说说图片的功能：
>- 首先随着鼠标的移动，收益曲线会有小圆圈跟着动，表示当前所在的位置。
>- 鼠标移动时会有一个提示框跟着动，显示当前的收益和时间。
>- 当鼠标移到曲线上方时，曲线会加粗，移到下方时则会正常

```python
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
matplotlib.rcParams['font.sans-serif'] = ['SimHei'] #指定默认字体
matplotlib.rcParams['axes.unicode_minus'] = False

import datetime
import numpy as np
import pandas as pd
import talib as ta
import tushare as ts
__color_grid__ ='#F0F0F4'
__color_axis__ = '#E9E7Ef'
__color_blue__ = '#6699FF'
```

因为本地跑，这里就不用优矿的API了，直接用tushare获取数据吧。

```python
quotes = ts.get_hist_data('600000', start='2015-06-01', end='2016-06-01')
quotes['tradeDate'] = quotes.index
quotes['tradeDate'] = quotes['tradeDate'].map (lambda x: mdates.date2num (datetime.datetime.strptime (x, '%Y-%m-%d')))
quotes['return'] = quotes['close']/quotes.ix[0, 'close'] -1
quotes.sort_values(by = 'tradeDate', ascending = True, inplace=True)
```

```python
fig = plt.figure(figsize=(10, 4), facecolor='w', edgecolor='w')
fig.set_tight_layout(True)
ax = fig.add_axes([0.05, 0.2, 0.9, 0.7])
ax.set_axisbelow(True)

ax.grid(True, color=__color_grid__, ls='-', lw=2, axis='y')
ax.set_ylim(np.min(quotes['return'])-0.5, np.max(quotes['return'])+0.5)
ax.set_xlim(-1, len(quotes) + 1)
for i in range(len(ax.get_children())):
    axis = ax.get_children()[i]
    if i != 2:
        if isinstance(axis, matplotlib.spines.Spine):
            axis.set_color('w')
    else:
        axis.set_color(__color_axis__)
        axis.set_lw(2)


ax.set_xticks(range(0, len(quotes), 30))
s1 = ax.set_xticklabels([mdates.num2date(quotes.ix[index, 'tradeDate']).strftime ('%Y-%m') for index in ax.get_xticks()])
```

```python
for i in range(len(ax.get_children())):
    axis = ax.get_children()[i]
    if i != 2:
        if isinstance(axis, matplotlib.spines.Spine):
            axis.set_color('w')
    else:
        axis.set_color(__color_axis__)
        axis.set_lw(2)


ax.set_xticks(range(0, len(quotes), 30))
s1 = ax.set_xticklabels([mdates.num2date(quotes.ix[index, 'tradeDate']).strftime ('%Y-%m') for index in ax.get_xticks()])
# s2 = ax.plot(np.array(quotes['return']), 'b-', alpha=1, lw=2, color=__color_blue__)


class Line(object):
    x = 0
    y = 0
    text_template = u'时间: %s\n策略: %0.2f'
    __color_blue__ = '#0033FF'
    xoffset, yoffset = -20, 20
    def __init__(self, ax, data):
        self.ax = ax
        self.data = data
        self.line = self.ax.plot(np.array(quotes['return']), 'b-', alpha=1, lw=2, color=self.__color_blue__)
        self.cirle = self.ax.plot([0], [self.data.ix[0, 'return']], 'bo', lw=5)
        self.annotation = ax.annotate (self.text_template,
                                       xy=(self.x, self.y), xytext=(self.xoffset, self.yoffset),
                                       textcoords='offset points', ha='left', va='top',
                                       bbox=dict(boxstyle='round, pad=0.3', fc='w', ec='#0033FF', alpha=0.5))
        self.annotation.set_visible (True)
        self.change = []

    def __call__(self, event):
        self.event = event
        # xdata, ydata = event.artist.get_data()
        # self.x, self.y = xdata[event.ind], ydata[event.ind]
        if event.xdata <= len(self.data)-1 and event.ydata != None:
            self.x, self.y = int(np.round(event.xdata)), event.ydata
            self.ax.lines.remove(self.ax.lines[1])
            self.time = self.data.index[self.x]
            self.open = self.data.ix[self.x, 'return']
            # print(event.xdata, event.ydata)
            if self.x is not None:
                if self.x < self.ax.get_xticks()[-1]/2:
                    self.cirle = self.ax.plot ([self.x], [self.data.ix[self.x, 'return']], 'bo', lw=5)
                    if self.y >= self.data.ix[self.x, 'return']:
                        self.change.append(plt.setp(self.line, lw=3))
                        self.annotation.xy = self.x+10, self.y+0.1
                        self.annotation.set_text(self.text_template % (self.time, self.open))
                        self.annotation.set_visible (True)
                        event.canvas.draw()
                    else:
                        self.change.append(plt.setp(self.line, lw=2))
                        self.annotation.xy = self.x+10, self.y+0.1
                        self.annotation.set_text (self.text_template % (self.time, self.open))
                        self.annotation.set_visible (True)
                        event.canvas.draw()
                else:
                    self.cirle = self.ax.plot ([self.x], [self.data.ix[self.x, 'return']], 'bo', lw=5)
                    if self.y >= self.data.ix[self.x, 'return']:
                        self.change.append (plt.setp (self.line, lw=3))
                        self.annotation.xy = self.x + 10, self.y + 0.1
                        self.annotation.set_text (self.text_template % (self.time, self.open))
                        self.annotation.set_visible (True)
                        event.canvas.draw ()
                    else:
                        self.change.append (plt.setp (self.line, lw=2))
                        self.annotation.xy = self.x- 30, self.y + 0.1
                        self.annotation.set_text (self.text_template % (self.time, self.open))
                        self.annotation.set_visible (True)
                        event.canvas.draw()
fig.canvas.mpl_connect('motion_notify_event', Line(ax, quotes))
```
