## <font color=#0099ff> 输入股票代码 </font>

```python
univ = ['002087.XSHE', '600971.XSHG', '000952.XSHE', '600710.XSHG', '600586.XSHG', '600963.XSHG']
```

```python
import numpy as np
import pandas as pd
from CAL.PyCAL import *
cal = Calendar('China.SSE')
endDate = cal.advanceDate(Date.todaysDate(), '-1B', BizDayConvention.Preceding).toISO()
```

## <font color=#0099ff> 1. 看看公司近期，有没有违规 </font>

```python
sys_code = DataAPI.SysCodeGet(codeTypeID=u"20048", field=['valueCD', 'valueName'],pandas="1")
sys_code = sys_code.rename(columns={'valueCD':'lllegalType', 'valueName':'l_type'})

sys_code2 = DataAPI.SysCodeGet(codeTypeID=u"20049", field=['valueCD', 'valueName'],pandas="1")
sys_code2 = sys_code2.rename(columns={'valueCD':'lllegalSubject', 'valueName':'e_type'})

sys_code3 = DataAPI.SysCodeGet(codeTypeID=u"20050", field=['valueCD', 'valueName'],pandas="1")
sys_code3 = sys_code3.rename(columns={'valueCD':'association', 'valueName':'a_type'})

sys_code4 = DataAPI.SysCodeGet(codeTypeID=u"20051", field=['valueCD', 'valueName'],pandas="1")
sys_code4 = sys_code4.rename(columns={'valueCD':'punishmentType', 'valueName':'p_type'})

ev_ = DataAPI.EquCompIllegalGet(secID=univ, beginDate=u"20170101", endDate=endDate, field=["secID", "secShortName", "publishDate", "lllegalType", "lllegalSubject", "adminobject", "association", "event", "punishmentType", "punishmentMeasure", "punishmentAmount"], pandas="1")
ev_copy = ev_

ev_ = ev_.applymap(lambda x: str(x))
ev_df = pd.merge(ev_, sys_code, how='left')
ev_df = pd.merge(ev_df, sys_code2, how='left')
ev_df = pd.merge(ev_df, sys_code3, how='left')
ev_df = pd.merge(ev_df, sys_code4, how='left')
ev_df = ev_df.drop('lllegalType', axis=1)
ev_df = ev_df.drop('lllegalSubject', axis=1)
ev_df = ev_df.drop('association', axis=1)
ev_df = ev_df.drop('punishmentType', axis=1)

ev_df = ev_df.rename(columns={'l_type':'违规类型', 'e_type':'违规主体', 'adminobject':'处罚对象', 'a_type':'关系', 'p_type':'处分类型', 'punishmentMeasure':'处分措施', 'punishmentAmount':'处罚金额(万元)'})
ev_df.set_index(['publishDate','secShortName','违规主体','处分类型', '关系', '违规类型', '处罚金额(万元)'])
# ev_copy
```

## <font color=#0099ff> 2. 看看公司近期的业绩预告 </font>

```python
sys_code = DataAPI.SysCodeGet(codeTypeID=u"70001", field=['valueCD', 'valueName'], pandas="1")
sys_code = sys_code.rename(columns={'valueCD':'reportType', 'valueName':'r_type'})

sys_code2 = DataAPI.SysCodeGet(codeTypeID=u"70006", field=['valueCD', 'valueName'], pandas="1")
sys_code2 = sys_code2.rename(columns={'valueCD':'forecastType', 'valueName':'f_type'})

re_ = DataAPI.FdmtEfGet(secID=univ,reportType=u"",endDate=endDate,beginDate=u"20180101",publishDateEnd=u"",publishDateBegin=u"",forecastType="",field=['secID','publishDate','secShortName','reportType','forecastType','NIncAPChgrLL','NIncAPChgrUPL','expnIncAPLL','expnIncAPUPL','expEPSLL','expEPSUPL'],pandas="1").drop_duplicates('secID')

re_['forecastType'] = re_['forecastType'].map(lambda x: str(x))
re_df = pd.merge(re_, sys_code)
re_df = pd.merge(re_df, sys_code2, how='left')
re_df = re_df.drop('reportType', axis=1)
re_df = re_df.drop('forecastType', axis=1)

re_df = re_df.rename(columns={'NIncAPChgrLL':'净利润增减幅度下限(%)', 'NIncAPChgrUPL':'净利润增减幅度上限(%)', 'expnIncAPLL':'预期净利润下限', 'expnIncAPUPL':'预期净利润上限', 'expEPSLL':'预期每股收益下限', 'expEPSUPL':'预期每股收益上限'})
re_df.set_index(['publishDate','secShortName','r_type', 'f_type'])
```

## <font color=#0099ff> 3. 看看公司的因子数据 </font>

```python
ticker_sector_df = DataAPI.EquIndustryGet(secID=univ, industryVersionCD='010303', intoDate=endDate, field='secID,secShortName,industry,industryID1,industryName1')
sectors = ticker_sector_df[['secID','secShortName','industryID1','industryName1']]

mdata = pd.DataFrame()
for stk in univ:
    data = DataAPI.MktStockFactorsDateRangeGet(secID=stk, beginDate=endDate, endDate=endDate, field='secID,tradeDate,PB,PE')
    data = data[['secID','PB','PE']]
    mdata = pd.concat([data,mdata])
m_data = mdata.merge(sectors, on='secID', how='outer')
m_data = m_data.sort('industryID1')
m_data = m_data.set_index([str('industryID1')])
m_data
```
