最近一直将TB里面的策略往优矿里面写，但是遇到了这样几个问题。
* 如何进行5分钟线，15分钟线，30分钟线的回测。我们知道简单的把refresh\_rate修改成5，15，30是不行的， 因为使用get_symbol_history()所获得的数据仍然是1分钟线。
* TB策略库所使用的语言变量很多，十分繁琐，如何简化。

这里采用了面向对象的编程方法，使用了四个类，DataCenter， Trans， Signals, Trader。其中，DataCenter类负责生成K线数据，Trans处理主力合约切换时的情形，Signals用来生成交易信号，Trader用来交易。
这样我们把策略分成这样几个类来处理，使我们的条理更加清晰。

**几点说明**
* 为了加快我们运行的速度，DataCenter类中的数据存放在双向队列中，我们可以根据所需K线的根数往里添加，首次我们一次性将队列存满，之后每次往里加一项，避免数据重复的加载
* 在主力进行切换时，Trans类会将当前的仓位全部平掉,并将DataCenter中保存的前主力合约的数据清空
* 原文策略中是有止盈止损的，当前没有想好止盈止损的设计，所以Signals不产生止盈止损的交易信号（小伙伴们帮我想想怎么止盈止损啊）

## 策略的逻辑
---
策略参考了TB中的《基于市场强弱指标和动量的通道突破系统（Superman）》

Superman系统通过计算市场的力量和速度来寻找交易机会。它使用市场强度指标MS来判断市场是否强到可以买入或者弱到可以卖出，使用两个动量指标来判断市场的上升或下跌速度。

**MS的计算:**
* 取最近5根K线并计算每根K线和前一根K线收盘价相比的涨跌幅
* 分别计算5根K线的涨跌幅之和S及5根K线中收盘上涨的K线的涨幅之和SU以及收盘下跌的K线的跌幅之和SD
* 若S>0,则MS=S/SU\*100;若S<0,则MS=S/|SD|\*100

**两个动量指标的计算：**
* Momentum1 = （Close[1]- Close[4]）/4
* Momentum2 = （Close[4]- Close[8]）/4

**策略：**
* 市场强度指标高于95时，Momentum1大于0, Momentum2小于0时计划买入
* 市场强度指标低于-95时，Momentum1小于0, Momentum2大于0时计划卖出
