```python
import numpy as np
import pandas as pd
import datetime
 
def strtodatetime(datestr):      
    format='%Y-%m-%d'
    return datetime.datetime.strptime(datestr,format)

def datediff(beginDate,endDate):  
       
    bd=strtodatetime(beginDate )  
    ed=strtodatetime(endDate )      
    oneday=datetime.timedelta(days=1)  
    count=0
    while bd!=ed:  
        ed=ed-oneday  
        count+=1
    return count

data=DataAPI.EquGet(secID=u"",ticker=u"",equTypeCD=u"A",listStatusCD=u"L",field=['secID'],pandas="1")
universe=data['secID']
filtered_universe=[]
for i in universe:
    tmp=DataAPI.EquSharesFloatGet(secID=i,ticker=u"",beginFloatDate=u"20180925",endFloatDate=u"20190325",\
                                  beginDate=u"",endDate=u"",field=u"",pandas="1")
    tmp.sort(columns='floatDate',ascending=False)
    a=strtodatetime('2018-12-30')
    #b=strtodatetime('2019-04-30')
    if len(tmp['floatDate'])!=0 :
        c=strtodatetime(tmp['floatDate'][0])
        t1=abs(a-c)
    #t2=abs(c-b)
        delta=t1.days>100 
        if delta:
            filtered_universe.append(i)
    else:
        filtered_universe.append(i)
print(len(filtered_universe))#1681#前后三个月没有解禁的股票
#print(len(universe))#2270
out=pd.DataFrame(filtered_universe)
out.to_csv('filtered_universe.csv')
                 
#再次寻找符合送转复合利润增长大于0.5的
```

```python
import pandas as pd
import numpy as np
a=pd.read_csv('filtered_universe.csv')
pool=a['0']
newpool=[]
for i in pool:
    tmp1=DataAPI.MktStockFactorsOneDayGet(tradeDate=u"20181120",secID=i,ticker=u"",field=['EPS'],pandas="1")
    tmp2=DataAPI.MktStockFactorsOneDayGet(tradeDate=u"20171120",secID=i,ticker=u"",field=['EPS'],pandas="1")
    tmp3=DataAPI.MktStockFactorsOneDayGet(tradeDate=u"20171120",secID=i,ticker=u"",field=['EPS'],pandas="1")
    tmp1.fillna(0)
    tmp2.fillna(0)
    tmp3.fillna(0)
    if len(tmp1)!=0 and len(tmp2)!=0 and len(tmp3)!=0:
        if tmp1['EPS'][0]>1 and tmp2['EPS'][0]>1 and tmp3['EPS'][0]>1:
            newpool.append(i)

#print(len(newpool))    #161
secondpool=[]
for i in newpool:
    tmp1=DataAPI.MktStockFactorsOneDayGet(tradeDate=u"20181120",secID=i,ticker=u"",field=['NetProfitGrowRate'],pandas="1")
    tmp2=DataAPI.MktStockFactorsOneDayGet(tradeDate=u"20171120",secID=i,ticker=u"",field=['NetProfitGrowRate'],pandas="1")
    tmp1.fillna(0)
    tmp2.fillna(0)
    if len(tmp1)!=0 and len(tmp2)!=0:
        if tmp1['NetProfitGrowRate'][0]>0 and tmp2['NetProfitGrowRate'][0]>0:
            secondpool.append(i)
print(len(secondpool))
out=pd.DataFrame(secondpool)
out.to_csv('final_filteredShares.csv')
```

```python
import pandas as pd
import numpy as np

data=pd.read_csv('final_filteredShares.csv')

def strtodatetime(datestr):      
    format='%Y-%m-%d'
    return datetime.datetime.strptime(datestr,format)
new=strtodatetime('2016-04-30')

pool=data['0']
selected=[]
for i in pool:
    tmp=DataAPI.EquDivGet(secID=i,ticker=u"",eventProcessCD=u"",exDivDate="",beginDate=u"20170101",endDate=u"20171231",\
                          beginPublishDate=u"",endPublishDate=u"",beginRecordDate=u"",endRecordDate=u"",field=['perShareDivRatio','perShareTransRatio'],pandas="1")
    tmp=tmp.fillna(0)
    
    if (len(tmp)!=0):
        ttt=tmp['perShareDivRatio'][0]+tmp['perShareTransRatio'][0]
        if (ttt<0.4):
            tmp2=DataAPI.EquIPOGet(secID=i,ticker=u"",eventProcessCD=u"",beginDate=u"",endDate=u"",field=['listDate'],pandas="1")
            isnew= strtodatetime(tmp2['listDate'][0])-new
            if isnew.days>0:
                selected.append(i)
    

#print(selected)
out={'secID':[],'secShortName':[]}        
for i in selected:
    tmp=DataAPI.EquGet(secID=i,ticker=u"",equTypeCD=u"A",listStatusCD=u"",field=['secID','secShortName'],pandas="1")
    out.get('secID').append(i)
    out.get('secShortName').append(tmp['secShortName'][0])

out=pd.DataFrame(out)
print(out)
#out.to_csv('potential_Div.csv')
    
```
