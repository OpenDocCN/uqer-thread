```python
import pandas as pd
import numpy as np
nameStr = 'a95_000906'                   ##因子信号文件。该文件为csv。共3列，列名依次为symbol、date、sig。symbol在csv文件中会自动去除代码前边的0；date为string格式，例如2011-01-15；sig为因子信号。
benchmark = 'HS300'                        # 策略参考基准
benchmarkcode = '000300'                  #中性化需要给定股票池。需要与策略参考基准一致

###交易日获取
trading_date = DataAPI.TradeCalGet(exchangeCD=u"XSHG,XSHE",beginDate=u"20120101",endDate=u"20151231",field=u"",pandas="1")
trading_date = trading_date[trading_date.isOpen==1].loc[:,['calendarDate']]
trading_date.rename(columns={'calendarDate':'date'},inplace=True)
trading_date = trading_date.drop_duplicates(subset='date')
###交易日的格式转换
trading_date['date_simple'] = trading_date.date.apply(lambda x: x[0:4]+x[5:7]+x[8:10])

###各交易日的动态股票池计算
universe_0 = pd.DataFrame(columns=['symbol','date'])
for i in range(len(trading_date)):
    universe = DataAPI.IdxConsGet(secID=u"",ticker=benchmarkcode, intoDate=np.array(trading_date.date_simple)[i], isNew=u"",field=u"",pandas="1")
    universe = universe.loc[:,['consID','intoDate']]
    universe.rename(columns={'consID':'symbol','intoDate':'date'},inplace=True)
    universe.date = np.array(trading_date.date)[i]
    universe_0 = pd.concat([universe_0, universe], axis=0)
universe_0 = universe_0.reset_index(drop=True)

####上传的因子文件数据（全A股，或者全zz800）
data = pd.read_csv( nameStr + '.csv')
data['symbol'] = data['symbol'].apply(lambda x: ((6-len(str(x)))*'0'+str(x)))
data['symbol'] = data['symbol'].apply(lambda x: x+'.XSHE' if (int(x))<500000 else x+'.XSHG' )

####与benchmark一致的动态因子池股票的选出
data_new = pd.merge(universe_0, data, on=['symbol','date'], how='left')
data_new = data_new.dropna()
data_new = data_new.reset_index(drop=True)
data_new = data_new.sort_values(by=['date','symbol'])
data_new = data_new.set_index('symbol')
data_test = data_new.copy()

#########因子的去极值
data_new['sig_winsorize'] = data_new.groupby('date')['sig'].apply(winsorize)
#########中性化
sig_neutralize_0 = []
for i in range(len(trading_date)):
    temp = data_new.loc[data_new.date==np.array(trading_date.date)[i], 'sig_winsorize']
    sig_neutralize = neutralize(temp, target_date=np.array(trading_date.date_simple)[i], risk_module='day', industry_type='SW1', exclude_style_list=[])
    sig_neutralize_0 = sig_neutralize_0 + list(sig_neutralize)
data_new['sig_winsorize_neutralize'] = sig_neutralize_0
#########标准化
data_new['sig_winsorize_neutralize_standardize'] = data_new.groupby('date')['sig_winsorize_neutralize'].apply(standardize)
########新因子的输出存储（保存为所读入的因子文件格式相同）
data_output = data_new.loc[:,['date','sig_winsorize_neutralize_standardize']]
data_output.rename(columns={'sig_winsorize_neutralize_standardize':'sig'},inplace=True)
data_output = data_output.reset_index(drop=True)
data_output['symbol'] = data_output['symbol'].apply(lambda x: x[0:6])
data_output.to_csv( nameStr+ '_' + benchmark + '_.csv',index=False )


#########因子的去极值、中性化、标准化  计算结果验证。可以选择某一天的因子计算，与优矿自带工具的计算做对比。
# print(data_new.loc[data_new.date=='2012-01-04',['sig_winsorize_neutralize_standardize']])
# data_test = data_test.loc[data_test.date=='2012-01-04',['sig']]
# data_test1 = winsorize(data_test.sig)
# data_test2 = neutralize(raw_data=data_test1, target_date='20120104', risk_module='day', industry_type='SW1', exclude_style_list=[])
# data_test3 = standardize(data_test2)
# print(data_test1)
# print(data_test2)
# print(data_test3)



```
