首先简单介绍一下“劲道指标”。劲道指标设计三个变量：当日收盘价、前日收盘价、当日成交量。计算公式：```劲道指标=当日成交量×(当日收盘价-前日收盘价)```。鉴于大多数技术分析指标要么只重视价格，要么只重视成交量的现状，up主我对埃尔德先生发明的这种将价格和成交量联系在一起的指标实在是一见倾心。在我看来埃尔德先生的交易思想为右侧交易，但是在入场点位的选择上却带着左侧交易的味道。在本策略中，劲道指标与5日均线配合作为买卖依据。由于均线具有提示股价运行方向的作用，因此只有在5日均线向上运行时，才会依据劲道指标选择入场时机。_当均线健康，劲道指标由正转负时，建仓；反之，均线不健康，劲道指标由负转正时，平仓。_
