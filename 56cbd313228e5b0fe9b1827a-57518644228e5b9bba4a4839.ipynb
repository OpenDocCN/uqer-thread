## 基于logistic回归的多因素选股策略
### 思路说明
#### 本策略是国信证券2010年研报 多因素模型及其在沪深300中的实证的第二篇 ，两篇报告和一个ppt[多因素选股研报](http://pan.baidu.com/s/1mgWYmve)，其实思路与第一篇的区别就在于APT模型的回归方法，第一篇要预测收益率，但是,预期收益率的确切估计是件比较困难的事情,同时,在投资中的很多策略也许并不要求获取较为确切的预期收益率数据,比如公募基金跑赢跟踪基准的要求以及结合股指期货的 alpha 策略,它们要求的是所选择的股票未来跑赢基准的概率有多大,这一要求比预测股票未来的预期收益率大为降低。而 logistic 模型正是解决 这一问题的利器。
#### 据说通联是已经为用户做了对冲，那么这个策略是不是会比传统的回归方法要好呢。
#### 我参考了研报的方法，在这里进行了因子极值处理，还请高手指点。
#### 最后说下问题，1，还是因子选择的问题，这个还要慢慢来选，包括之前孙老师那篇互联网大数据选股中用到的新闻热度等数据可以加进来。2，其实Logistic模型只是给出了因子与战胜基准概率之间的定量关系,至于他们之间的逻辑、因子与因子之间的关联则研究的很少,没有合理的逻辑,那么就无法保证未来选股结果的可靠性和稳定性，而且基于机器学习中的训练数据只是上一期的数据，测试数据是本期数据这一点我还是有点疑虑。总之吧，思想最重要，方法只是招式，我们还是不要轻易乱套模型了。


```python
def StockFactorsGet(trading_day,stock):
    data =  DataAPI.MktStockFactorsOneDayGet(tradeDate=trading_day.strftime('%Y%m%d'),secID=stock,field=['secID','tradeDate']+used_factors,pandas="1")
    data['ticker'] = data['secID'].apply(lambda x: x[0:6])
    data.set_index('ticker',inplace=True)
    #数据标准化
    for f in used_factors:
        if data[f].std() == 0:
            continue
        data[f] = (data[f] - data[f].mean()) / data[f].std()
    return data
```
