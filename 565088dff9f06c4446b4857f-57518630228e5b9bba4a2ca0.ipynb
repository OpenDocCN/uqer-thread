# 测量系统指标：R立方

UQER 提供了很多指标来考察一个系统的好坏，包括 sharp ratio, alpha, beta 值等，而在海龟交易法则里提到过：

测试时不要忽略样本对总体的代表性。

如果对数据稍作改动，不会影响一个统计指标，我们就说这个指标是稳健的。
对数据变化太敏感的指标都不够稳健。

其中稳健的指标有：
1，回归年度回报率（regressed annual return：RAR）
2，R立方（robust risk/reward ratio）
R立方=RAR/平均最大衰落

而UQER里提供的 sharp ratio, 以及常用的 CAGR、Mar ratio 等都不能算作稳健的指标，下面是R立方指标的实现，供大家使用交流。
海龟原文里是使用年度回报率的回归，鉴于大家多数都不去跑个10年20年的数据，在这里我把参数设置成30天(days = 30)，然后做了年化，原文里长度调整是除以365，在UQER里的bt数据都是交易日，所以我这里用350天来代替(days_of_year = 350)

使用的方式就是在运行完一个算法的回测之后，新建一个code单元，把代码粘进去，选 运行->运行(ctrl+enter)

```python
from sklearn import linear_model
from numpy.lib.stride_tricks import as_strided
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import operator

days_of_year = 350
days = 30
top_drawdown = 5

def cum_drawdown(values, start, end, limit):
    window_values = values[start:end]
    i = np.argmax(np.maximum.accumulate(window_values) - window_values) # end of the period
    if i == 0:
        return []
    j = np.argmax(window_values[:i]) # start of period
    drop = [{ "drawdown":window_values[j]- window_values[i], "length":i-j}]
    if limit > 0:
        if j > 0:
            drop += cum_drawdown(values, start, j+start, limit - 1)
        if i+start < end-1:
            drop += cum_drawdown(values, i+start, end, limit - 1)
    return sorted(drop, key=lambda x:x.get("drawdown"), reverse=True)[:limit]

clf = linear_model.LinearRegression()
portfolio_return = (bt.portfolio_value.diff()/(bt.portfolio_value - bt.portfolio_value.diff()))
#计算 RAR
X = [[t] for t in bt.index.tolist() if t % days == 0]
Y = [np.sum(portfolio_return[(t[0]-days):t[0]]) for t in X]
clf.fit(X,Y)
regressed_annual_return = clf.coef_[0] * (days_of_year / days)
#计算带时间长度调整的前 top_drawdown 次最大衰落
drawdown = cumdrawdown(portfolio_return.dropna().tolist(), 0, len(bt.portfolio_value.tolist()), top_drawdown)
drawdown_mean = np.mean([x['drawdown'] for x in drawdown])
drawdown_length_mean = np.mean([x['length'] for x in drawdown])
plt.plot(X,Y)
#输出R^3指标
print("%s days of R3: %s" % (days, regressed_annual_return/(drawdown_mean * drawdown_length_mean/days_of_year)))
```
