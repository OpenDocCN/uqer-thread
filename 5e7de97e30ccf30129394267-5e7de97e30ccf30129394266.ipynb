# 上证50因子序列

```python
import numpy as np
import pandas as pd
import datetime
```

**计算回归权重**

```python
factor_list=['CMRA','DAVOL5','DVRAT','LCAP','MA10','REVS10','PB','PE']
index_code = 'SH50'
begin_Date = '20190901'
end_Date = '20190930'


def factor_series(equity_code,begin_Date,end_Date):#获取每只股票一段时间因子序列
    temp_df = DataAPI.MktStockFactorsDateRangeGet(secID=equity_code,beginDate=begin_Date,endDate=end_Date,field=['secID','tradeDate']+factor_list,pandas="1")
    return temp_df

def get_all_factor(index_code,begin_date,end_date):
    equity_pool = set_universe(index_code,end_date)
    df = pd.DataFrame()
    for i in equity_pool:
        temp_df = factor_series(i,begin_date,end_date)
        df=pd.concat([df,temp_df])
    df.fillna(method = 'ffill',inplace = True)
    return df

def industry_pool(index_code,date):#多只股票某一天的行业因子值
    industry_list = ['secID','secShortName','industryName1']
    equity_pool = set_universe(index_code,date)
    Industry_factor = DataAPI.EquIndustryGet(secID=equity_pool,industryVersionCD="010303",
                                intoDate=date,field=industry_list,pandas="1") #固定的
    Industry = pd.merge(Industry_factor,pd.get_dummies(Industry_factor.industryName1),left_index = True,right_index = True)
    del Industry['industryName1']
    Industry['cons'] = 1
    return Industry

def concat_df():
    a = get_all_factor(index_code,begin_Date,end_Date)
    d = industry_pool(index_code,end_Date)
    new_df = pd.merge(a,d,on = 'secID',how = 'inner')
    return new_df,new_df.tradeDate

def generate_vec(dfs,Industry_name):
    ratio = np.zeros(len(Industry_name +factor_list)-1)
    for i in range(1,len(Industry_name)-1):
        ratio[i] = -(dfs.weight[dfs[Industry_name[i]]==1].sum()/np.sum(dfs.weight))/(dfs.weight[dfs[Industry_name[-1]]==1].sum()/np.sum(dfs.weight))
    return ratio

def regression(df):
    #进行回归
    d = industry_pool(index_code,end_Date)
    Industry_name = ['cons']+[str(i) for i in d.columns[2:-1]]
    W = np.diag(np.sqrt(df.weight)/np.sum(np.sqrt(df.weight)))
    X = df[Industry_name +factor_list].values
    R = np.zeros((len(Industry_name +factor_list),len(Industry_name +factor_list)-1))
    R[len(Industry_name)-1,:] = generate_vec(df,Industry_name)
    for i in range(len(Industry_name)-1):
        R[i,i] = 1
    for i in range(len(Industry_name),len(Industry_name+factor_list)):
        R[i,i-1] = 1
    M = X.dot(R)
    S = (M.T).dot(W).dot(M)
    O = R.dot(np.linalg.inv(S)).dot(M.T).dot(W)
    return O.T

def wrangle_df(df,str_list):
    O = pd.DataFrame()
    for i in str_list:
        temp_df = df[df['tradeDate']==i]
        #normalize化
        market_value = np.exp(temp_df.LCAP)
        weight = market_value/np.sum(market_value)
        temp_df[factor_list] = temp_df[factor_list].apply(lambda x:(x-x.dot(weight))/np.std(x))
        temp_df['weight'] = weight
        temp_O = pd.DataFrame(regression(temp_df))  
        temp_O['date'] = i
        temp_O['code'] = temp_df.secID.values
        temp_O.set_index(['date','code'],inplace=True)
        temp_O['weight'] = weight.values
        rf=DataAPI.MktIborGet(ticker=u"Shibor1D",tradeDate=i[:4]+i[5:7]+i[8:10],field="rate")/252  #当日无风险收益
        Return = DataAPI.EquRetudGet(secID=list(temp_df.secID),beginDate=i[:4]+i[5:7]+i[8:10],endDate=i[:4]+i[5:7]+i[8:10],field=
                                     ['secID',"dailyReturnNoReinv"],pandas="1")
        y = Return['dailyReturnNoReinv'].values-rf.values[0,0]
        temp_O['excess_rf'] = y
        O = pd.concat([O,temp_O])
        print(i)
    return O

def main():
    d = industry_pool(index_code,end_Date)
    Industry_name = ['cons']+[str(i) for i in d.columns[2:-1]]
    df,date_list = concat_df()
    e=wrangle_df(df,date_list.unique())
    e.columns = ['市场因子']+[i+'行业因子' for i in Industry_name[1:]]+[i+'因子' for i in factor_list]+['weight','excess_rf']
    return e

factor_weights = main()
factor_weights 
```
