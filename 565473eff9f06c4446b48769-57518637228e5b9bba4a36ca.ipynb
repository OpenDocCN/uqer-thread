# 基金产品的简单回测框架（account类的定义）

最近比较关注基金，投资基金可以不用太过强调风险的分散，而且主要涉及到的净值也相对好分析，于是一直想试着用uqer的平台来做基金策略的回测，不过uqer好像只支持股票和特定基金的回测，于是就仿照uqer写了一个简单的基金回测框架，这个主要是其中account类的定义，抛砖引玉，希望能和对基金回测感兴趣的朋友们有些沟通：）

```python
import pandas as pd
import lib.get_data
import lib.workday

front_fee_rate=0.006
end_fee_rate=0.01

class account:

    # data = get_fund_nav(self.start_date, self.end_date)
    
    def __init__(self, cash):
    # 初始化起始日期、结束日期、初始现金
        self.current_date = ''    # 当前日期
        self.cash = float(cash)
        # self.fund_position = pd.DataFrame({'fund_ticker':['000001','000003'], 'position':[200,300]},columns=['fund_ticker', 'position'])
        self.fund_position = {}    # 基金代码及头寸，dict类型
        # self.asset = self.cash + sum(self.fund_position.position)*(1-end_fee_rate)
        self.asset = self.cash + sum(self.fund_position[x] for x in self.fund_position.keys())*(1-end_fee_rate)    # 计算总资产
    
    def set_current_date(self, date):
        self.current_date = lib.workday.current_date(date)
        
    def sell(self, ticker, amount):
    # 卖出特定基金，卖出数量高于现有数量时全部卖出
        amount = float(amount)
        current_amount = self.fund_position[ticker]
        if amount < current_amount:
            self.fund_position[ticker] -= amount
            self.cash += amount*(1-end_fee_rate)
        else:
            self.cash += current_amount*(1-end_fee_rate)
            del self.fund_position[ticker]
        self.asset = self.cash + sum(self.fund_position[x] for x in self.fund_position.keys())*(1-end_fee_rate)
            
    def buy(self, ticker, amount):
    # 买入特定基金，买入数量高于现金时将现金用完
        if self.cash > 0.01:
            amount = float(amount)
            if amount>self.cash:
                if ticker in self.fund_position:
                    self.fund_position[ticker] += self.cash/(1+front_fee_rate)
                else:
                    self.fund_position[ticker] = self.cash/(1+front_fee_rate)
                self.cash = 0.0
            else:
                if ticker in self.fund_position:
                    self.fund_position[ticker] += amount/(1+front_fee_rate)
                else:
                    self.fund_position[ticker] = amount/(1+front_fee_rate)
                self.cash -= amount
        else:
            self.cash = 0.0
        self.asset = self.cash + sum(self.fund_position[x] for x in self.fund_position.keys())*(1-end_fee_rate)

    def update(self):
    # 更新每日资产
        today = self.current_date
        yesterday = lib.workday.last_day(today)
        for ticker in self.fund_position:
            nav =  lib.get_data.get_ticker_nav(ticker, start_date=yesterday, end_date=today)
            self.fund_position[ticker] *= nav.ADJUST_NAV[1]/nav.ADJUST_NAV[0]
        self.asset = self.cash + sum(self.fund_position[x] for x in self.fund_position.keys())*(1-end_fee_rate)
    
    def handle_data(self):
        # 简单测试
        self.buy('000001', 100)

# 测试
if __name__ == '__main__':
    lc_account = account(1000)
    lc_account.set_current_date('20151007')
    # print lc_account.fund_position
    # print lc_account.asset
    lc_account.buy('000001',1000)
    print lc_account.current_date
    print lc_account.fund_position
    print lc_account.asset
    workday = lib.workday.date_between('20151008','20151110')
    for date in workday:
        lc_account.set_current_date(date)
        lc_account.update()
        print lc_account.current_date
        print lc_account.fund_position
        print lc_account.asset
    # lc_account.sell('000001',10000)
    # print lc_account.fund_position
    # print lc_account.asset
    
```
