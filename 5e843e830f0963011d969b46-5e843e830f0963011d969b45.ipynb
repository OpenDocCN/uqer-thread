## 基础函数，转换数据格式

```python
# 转换stockcode代码为str类型
def code2str(x):
    x_str = str(x)
    while int(x)<100000:
        x = x*10
        x_str = '0' + x_str
    return x_str
```

```python
original_df = pd.read_csv('original_data.csv',encoding='GBK')
original_df['stockcode'] = original_df['stockcode'].apply(code2str)
```

```python
# 再次读取文件之后，是一个整体的str，转化为str构成的list
# 每个ykcode的长度为11，中间还有一个逗号和一个空格,首尾有[],但是缺少逗号和空格
def ykcodestr2strlist(x):
    return [x[1+i*11+i*2:1+(i+1)*11+i*2] for i in range(int(len(x)/13))]

# 再次运行时，从ykcode_by_tradeDate.csv中，ykcode会成为一整个字符串，要重新编排格式
ykcode_by_tradeDate_df = pd.read_csv('ykcode_by_tradeDate.csv', encoding='gbk')
ykcode_by_tradeDate_df['ykcode_list'] = ykcode_by_tradeDate_df.ykcode_list.apply(lambda x: ykcodestr2strlist(x))
```

```python
# 从文件中读取stockcode_by_tradeDate，要把stockcode的一整个str转化为由str组成的list
# stockcode长度为2，仍旧有空格和逗号
def stockcodestr2list(x):
    return [x[1+i*6+i*2:1+(i+1)*6+i*2] for i in range(int(len(x)/8))]

# 从文件中读取stockcode
stockcode_by_tradeDate_df = pd.read_csv('stockcode_by_tradeDate.csv', encoding='gbk')
stockcode_by_tradeDate_df['stockcode_list'] = stockcode_by_tradeDate_df.stockcode_list.apply(lambda x: stockcodestr2list(x))
```

# ifoperateRato_con	
近2期年报扣非净利润（扣除非经常性损益后的净利润）均小于0或者最新财报与上一年年报扣非净利润均小于0
DataAPI.FdmtEeGet
获取扣非净利润，Q1-第一季报，S1-半年报，Q3-第三季报，CQ3-三季报（累计1-9月），A-年报。
NIncomeCut	float	扣除非经常性损益后净利润
用np.isnan()来判断是否为空！！！

```python
FdmtEeGet_df = DataAPI.FdmtEeGet(ticker=ticker,secID=u"",reportType=u''
                                 ,endDate=end_date,beginDate=begin_date,publishDateEnd=u"",publishDateBegin=u"",
                                 field=Q_basic_factor+['NIncomeCut','reportType'],pandas="1")
new_df = FdmtEeGet_df[pd.isnull(FdmtEeGet_df.NIncomeCut) == False]
new_df.reset_index(drop=True,inplace=True)
new_df['publishDate'] = new_df.publishDate.apply(lambda x: x[:10])
# 将新产生的指标放在记入新列，以publishDate为关键词
# 都是要判断上一年的'年报'是否小于0
def ifoperateRato_con(x):
    if np.isnan(x.NIncomeCut):
        return np.NAN
    elif x.NIncomeCut<0:
        last_year = str(int(x.endDate[:4])-1)+'-12-31' # 转换成上一年的年报endDate 
        temp1 = new_df.ticker == x.ticker 
        temp2 = new_df.endDate == last_year # 有不同的publishDate对应同一个endDate
        temp3 = new_df.NIncomeCut[temp1 & temp2]
        if len(temp3)==0:
            # print('nan')
            return np.NAN
        elif True in list(temp3<0):
            # print('NIC:'+str(x.NIncomeCut))
            # print('1: '+str(temp3))
            return True # 连续两年都小于0
        else:
            # print('NIC:'+str(x.NIncomeCut))
            # print('0: '+str(temp3))
            return False # 上一年不小于0
    else:
        return False # 本年不小于0

new_df['ifoperateRato_con'] = new_df.apply(lambda x: ifoperateRato_con(x),axis = 1)
new_df=new_df[['ticker','publishDate','endDate','ifoperateRato_con']]
# 因为publishDate不一致，以及数据量少，按照endDate合并
new_df.drop_duplicates(inplace=True)
new_df.reset_index(drop=True,inplace=True)
new_df.to_csv('ifoperateRato_con.csv', encoding='GBK',index=False)
```

# ifoperatecash_cond
最近一期财报及上2期年报经营性现金流净额和小于0（除银行、非银金融、房地产）

```python
# NCFOperateA	float	经营活动产生的现金流量净额
FdmtCFGet = DataAPI.FdmtCFGet(ticker=ticker,secID=u"",reportType=u"",
                        endDate=end_date,beginDate=begin_date,publishDateEnd=u"",publishDateBegin=u"",endDateRep="",
                              beginDateRep="",beginYear="",endYear="",fiscalPeriod="",
                              field=Q_basic_factor+['NCFOperateA','reportType'],pandas="1")
# 有publishdate不同，但是endDate相同的年报。。。只要有一个true就true吧
# 可能为空的值也要考虑
def ifoperatecash(x):
    if x.NCFOperateA > 0:
        return False
    else:
        last_year1 = str(int(x.endDate[:4])-1) + '-12-31'
        last_year2 = str(int(x.endDate[:4])-2) + '-12-31'
        temp1 = (FdmtCFGet.endDate == last_year1) & (FdmtCFGet.reportType == 'A') & (FdmtCFGet.ticker == x.ticker)
        temp2 = (FdmtCFGet.endDate == last_year2) & (FdmtCFGet.reportType == 'A') & (FdmtCFGet.ticker == x.ticker)
        temp3 = FdmtCFGet.NCFOperateA[temp1]
        temp4 = FdmtCFGet.NCFOperateA[temp2]
        if (len(temp3) == 0) | (len(temp4) == 0):
            return np.NAN
        else:
            return (True in list(temp3<0)) & (True in list(temp4<0)) # 要转化为list才可以查找true是否在其中
        
FdmtCFGet['ifoperatecash_cond'] = FdmtCFGet.apply(lambda x: ifoperatecash(x),axis = 1)
FdmtCFGet = FdmtCFGet[['ticker','publishDate','endDate','ifoperatecash_cond']]
FdmtCFGet.to_csv('ifoperatecash_cond.csv', index=False, encoding='GBK')
```

# ifmny_debtRatio_cond	
货币资金/1年内负债覆盖倍数小于0.2（不包括银行、非银金融，企业性质不含央企、国企）
# mny_debtRatio	
货币资金/1年内负债覆盖倍数

```python
# 货币资金
# NCAWithin1Y	float	一年内到期的非流动资产
# cashCEquiv	float	货币资金
field_list = Q_basic_factor+['tradingFL','STBorr','NotesPayable','othCL','NCAWithin1Y','cashCEquiv']
FdmtBSGet = DataAPI.FdmtBSGet(ticker=ticker,secID=u"",reportType=u"",endDate=end_date,beginDate=begin_date,
                              publishDateEnd=u"",publishDateBegin=u"",endDateRep="",beginDateRep="",
                              beginYear="",endYear="",fiscalPeriod="",
                              field=field_list,pandas="1")
FdmtBSGet['loan_sum'] = FdmtBSGet[['tradingFL','STBorr','NotesPayable','othCL','NCAWithin1Y']].sum(axis = 1)
def fun_ifmny_debtRatio(x):
    if np.isnan(x.cashCEquiv) | np.isnan(x.loan_sum):
        # print('nan')
        return np.NAN
    elif int(x.loan_sum) < 0.0001:
        return np.NAN
    else:
        return x.cashCEquiv/x.loan_sum < 0.2
FdmtBSGet['ifmny_debtRatio_cond'] = FdmtBSGet.apply(lambda x: fun_ifmny_debtRatio(x), axis = 1)
def fun_mny_debtRatio(x):
    if np.isnan(x.cashCEquiv) | np.isnan(x.loan_sum):
        # print('nan')
        return np.NAN
    elif int(x.loan_sum) < 0.0001:
        return np.NAN
    else: 
        return x.cashCEquiv/x.loan_sum
FdmtBSGet['mny_debtRatio'] = FdmtBSGet.apply(lambda x: fun_mny_debtRatio(x), axis = 1)
FdmtBSGet = FdmtBSGet[['publishDate','endDate','ticker','ifmny_debtRatio_cond','mny_debtRatio']]
FdmtBSGet.to_csv('mny_debtRatio.csv',encoding='gbk',index=False)
```

# ifholdRatio_cond	
最近一期财报及上一期年报机构持仓比例均低于5%（包括陆股通，信托公司除外）
按照endDate来group求和
按照endDate判断财报是否年报
这个publishDate可能不能去merge到季度因子中，因为披露时间是不同的
shName的长度大于3时，判断为机构股

```python
EquShTenGet = DataAPI.EquShTenGet(secID=u"",ticker=ticker,beginDate=begin_date,endDate=end_date,
                                  field=['ticker','endDate','shName','holdPct','publishDate'],pandas="1")

# 判断是否为机构持股,汉字编码是三个为一个
EquShTenGet['cond'] = EquShTenGet.shName.apply(lambda x: True if len(x)>9 else(False))
# 提取机构股
new_df = EquShTenGet.loc[EquShTenGet.cond == True]
# 按照endDate和publishDate求和,用transform来扩展
new_df['sum'] = new_df.groupby(['ticker','endDate','publishDate'])['holdPct'].transform('sum')
# 删除由于publishDate为nan造成的holdPct的nan
new_df = new_df.loc[pd.isnull(new_df['sum'])==False]
# 保留有效的key
new_df.drop(['shName','holdPct'],axis = 1,inplace=True)
new_df = new_df.drop_duplicates()
new_df.reset_index(drop = True, inplace=True)
new_df['ifholdRatio'] = new_df.apply(lambda x: x['sum'] < 5, axis =1)
# 最近一期财报及上一期年报机构持仓比例均低于5%
# 有publishdate不同，但是endDate相同的年报。。。只要有一个true就true吧
# 可能为空的值也要考虑
def fun_ifholdRatio(x,df):
    if x.ifholdRatio == False:
        return False
    else:
        last_year1 = str(int(x.endDate[:4])-1) + '-12-31'
        temp1 = (df.endDate == last_year1) & (df.ticker == x.ticker)
        temp3 = df.ifholdRatio[temp1]
        if len(temp3) == 0:
            return False
        else:
            return True in list(temp3.values) # 要转化为list才可以查找true是否在其中
new_df['ifholdRatio_cond'] = new_df.apply(lambda x: fun_ifholdRatio(x,new_df),axis = 1)
new_df = new_df[['ticker','endDate','ifholdRatio_cond']]
new_df.to_csv('ifholdRatio_cond.csv',encoding='gbk',index = False)
```

ROE_G_3y	因子profit_ttm（当日）除以tot_equity（最近一期已公布财报）三年复合增长率

```python
def cubic_root(nr):
    if nr<0:
        return -np.power(-nr, float(1)/3)
    else:
        return np.power(nr, float(1)/3)
```

```python
def cal_ttm(df,base_str,new_str):
    df[new_str] = df.apply(lambda x:df[(df.ticker==x.ticker)&(df.year==x.year-1)&(df.period==x.period)][base_str].values.tolist(),axis=1)
    df[new_str] = df[new_str].apply(lambda x:np.nan if len(x)==0 else x[-1])
    df[new_str] = ((df[base_str]-df[new_str])/df[new_str])
    return df

def cal_3y(df,base_str,new_str):
    df[new_str] = df.apply(lambda x:df[(df.ticker==x.ticker)&(df.year==x.year-3)&(df.period==x.period)][base_str].values.tolist(),axis=1)
    df[new_str] = df[new_str].apply(lambda x:np.nan if len(x)==0 else x[-1])
    df[new_str] = ((df[base_str]-df[new_str])/df[new_str]).apply(cubic_root)
    return df
```

# 合并季度因子

```python
# 首先根据债券存续期提取三大基础报表的有效数据，因为有的报表有一部分缺失，所以采用outer方法合并
# x是整体的值
# start_end_df 表示了开始和结束日期
def select_effective(x, start_end_df):
    temp1 = x.publishDate >= start_end_df.start_date
    temp2 = x.publishDate <= start_end_df.end_date
    temp3 = x.ticker == start_end_df.stockcode
    temp4 = start_end_df.loc[temp1 & temp2 & temp3]
    if len(temp4) == 0:
        return np.nan
    else:
        return temp4.ykcode.values[0]
```

```python
# 合并的时候重新读取数据，ticker为float形式，前面的0不要补齐，因为都没有
# 合并三大报表的基础季度因子
def basic_Q():
    temp_df = pd.read_csv('FdmtIndiGrowthGet.csv',encoding='gbk')
    start_end_df = pd.read_csv('start_end_df.csv',encoding='gbk')
    temp_df['ykcode']= temp_df.apply(lambda x: select_effective(x,start_end_df), axis = 1)
    temp_df = temp_df[pd.isnull(temp_df.ykcode)==False]

    temp_df1 = pd.read_csv('FdmtIndiRtnGet.csv',encoding='gbk')
    temp_df1['ykcode']= temp_df1.apply(lambda x: select_effective(x,start_end_df), axis = 1)
    temp_df1 = temp_df1[pd.isnull(temp_df1.ykcode)==False]

    temp_df2 = pd.read_csv('FdmtIndiTrnovrGet.csv',encoding='gbk')
    temp_df2['ykcode']= temp_df2.apply(lambda x: select_effective(x,start_end_df), axis = 1)
    temp_df2 = temp_df2[pd.isnull(temp_df2.ykcode)==False]

    key_list = ['secID','publishDate','endDate','partyID','ticker','ykcode']
    new_df = pd.merge(temp_df,temp_df1,left_on = key_list, right_on = key_list,how = 'outer')
    new_df = pd.merge(new_df,temp_df2,left_on = key_list,right_on=key_list,how = 'outer')
    new_df.reset_index(drop=True,inplace=True)
    new_df = pd.merge(new_df,start_end_df,left_on = ['ykcode'],right_on=['ykcode'],how = 'left')
    new_df.drop(['ticker'],inplace=True,axis = 1)
    return new_df
new_df = basic_Q()
new_df.to_csv('Q_temp.csv',encoding='gbk',index=False)
```

```python
def add_industry(new_df,filename,ind_type):
    industry_df = pd.read_csv(filename,encoding='gbk')
    # industry_df 行业标签
    indID = []

    new_df.reset_index(drop=True,inplace=True)
    for i in new_df.index:
        temp1 = industry_df.ticker.values==new_df.stockcode.values[i]
        temp2 = industry_df.intoDate <= new_df.publishDate[i]
        temp3 = industry_df.outDate > new_df.publishDate[i]
        # 存在提取不到时间的
        temp4 = industry_df.industryID1[temp1&temp2&temp3]
        if len(temp4) == 0:
            indID.append(np.NAN)
        else:
            indID.append(temp4.values[0])
    new_df[ind_type] = indID
    return new_df

def add_bond(new_df,filename,ind_type):
    industry_df = pd.read_csv(filename,encoding='gbk')
    # industry_df 行业标签
    indID = []

    new_df.reset_index(drop=True,inplace=True)
    for i in new_df.index:
        temp1 = industry_df.secID.values==new_df.ykcode.values[i]
        temp2 = industry_df.intoDate <= new_df.publishDate[i]
        temp3 = industry_df.outDate > new_df.publishDate[i]
        # 存在提取不到时间的
        temp4 = industry_df.rating[temp1&temp2&temp3]
        if len(temp4) == 0:
            indID.append(np.NAN)
        else:
            indID.append(temp4.values[0])
    new_df[ind_type] = indID
    return new_df
```
