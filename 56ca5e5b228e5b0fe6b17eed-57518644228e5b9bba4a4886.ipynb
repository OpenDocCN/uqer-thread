# 未来我为什么只关注上证50系列产品

这是我在优矿的第三篇文章，我深知上一篇写的好差，但是我为什么还放出来，自有我的道理，我不敢保证我以后不会再写狗屎一样的文章，但是我还是想更多的把我的心得告诉大家，也希望大家收到一点启发，虽然也许我说的都是错的。
好多朋友都可能会想，你都不贴策略，只是嘴皮子功夫。说实话，我不会写，但是我已经派了几个好基友潜伏在优矿了，也许你现在克隆的某个策略就承载着我的思想。
言归正传，这贴只谈理由，至于什么是上证50，自行百度，至于背景，以后有机会写到我自己个人经历的时候，也许你就明了了。
1，上证50产品是国内目前唯一的全频道产品，包括指数，etf，分级基金，期货，期权，将这些产品进行组合，不管是在上涨时，下跌时，甚至在平盘时都有赚钱的机会，你觉得可怕吗？
2，上证50指数中包括的股票都是质地好，市盈率低的股票，在目前的状态下，相对风险较小。
3，上证50指数期货和新加坡富时a50期货具备极强的正相关，而新加坡富时a50期货国际化程度很高，因此上证50将客观反映市场。
这就是我选择关注上证50产品的理由。
我觉得这就是干货。所以我拜托各位鲜花，拍砖，让我能实现我的愿望。
