# PB市净率因子研究

PB市净率作为常用的判断公司价格是否合理的股票投资基本分析最常见的参考指标之一，笔者通过阶段性（优矿+python）的学习，做一篇比较基础的研究，希望大家多多给予改进建议，在此先感谢！
PB因子研究步骤：
1、数据处理：将PB为负的个股PB改为所有股票PB中最大的数
2、测试：
    1）将PB由低到高分为4组，观察4组净值以及对冲后后收益的净值表现。
    2）将PB用中心化函数剔除行业因素后，再由低到高分为4组，观察4组净值以及对冲后后收益的净值表现。
3、比较：
    1）比较分位数表现是否PB低的>PB高的组合
    2）比较剔除行业因素的中心化组合是否好于没有中心化的组合

```python
import matplotlib.pyplot as plt
from matplotlib import rc
rc('mathtext', default='regular')
import seaborn as sns
sns.set_style('white')
from matplotlib import dates
import datetime
import numpy as np
import pandas as pd
import time
import scipy.stats as st
from CAL.PyCAL import *    # CAL.PyCAL中包含font
cal=Calendar('China.SSE')
```

```python
start = '2010-01-01'                       # 回测起始时间
end = '2017-01-01'                         # 回测结束时间
benchmark = 'HS300'                        # 策略参考标准
universe = set_universe('HS300')           # 证券池，支持股票和基金
capital_base = 1000000                     # 起始资金
freq = 'd'                                 # 策略类型，'d'表示日间策略使用日线回测，'m'表示日内策略使用分钟线回测
refresh_rate = Monthly(1)                  # 调仓频率，表示执行handle_data的时间间隔，若freq = 'd'时间间隔的单位为交易日，若freq = 'm'时间间隔为分钟

sim_params = quartz.SimulationParameters(start, end, benchmark, universe, capital_base)
data = quartz.get_backtest_data(sim_params)   #  获取回测区间所用基础行情数据
qtlist = [0,1,2,3]

bt_record = dict()
risk_record = dict()

for qt in qtlist:

    print "back testing: ", qt
    def initialize(account):                   # 初始化虚拟账户状态
        pass

    def handle_data(account):                  # 每个交易日的买入卖出指令
        close_price = account.reference_price
        factorlist = DataAPI.MktStockFactorsOneDayGet(tradeDate=account.previous_date.strftime("%Y%m%d"),secID=account.universe,field=u"secID,pb",pandas="1")#获取数据
        #数据处理
        maxPB = factorlist['PB'].max()
        for i in range(0,len(factorlist),1):
            if factorlist['PB'][i] < 0:
                factorlist['PB'][i] = maxPB
        #排序
        factorlist.set_index('secID', inplace = True)
        list = factorlist['PB'].dropna().to_dict()
        list = sorted(list.items(), lambda x, y: cmp(x[1], y[1]))
        #构建篮子
        size = len(list)
        buylist = []
        for i in range(size/4*qt,size/4*(qt+1)-1,1):
            buylist.append(list[i][0])
        #交易
        for s in  account.avail_security_position: # 卖出不在买股列表上的股票
            if s not in buylist:
                order_pct_to(s,0)
        for s in buylist:
            if not (np.isnan(close_price[s]) or (close_price[s] == 0)): # 买入列表的股票（等权买入）
                order_pct_to(s, 4.0/size)
        return
    
    strategy = quartz.TradingStrategy(initialize, handle_data)
    # 回测部分
    try:
        bt, perf = quartz.quick_backtest(sim_params, strategy, data=data, refresh_rate=refresh_rate)

        bt_record[qt] = bt
        risk_ret = dict()
        out_keys = ['annualized_return', 'volatility', 'information_ratio', 'sharpe', 'max_drawdown', 'alpha', 'beta']
        for key in out_keys:
            risk_ret[key] = perf[key]
            risk_ret['bt'] = bt
            risk_record[qt] = risk_ret
    except Exception, e:
        print e
        continue

pd.DataFrame(risk_record).T.sort('information_ratio')
            
```

从α来看，pb越高，α越大。但信息比率较低，波动看上去较大。

```python
fig = plt.figure(figsize=(20,16))
fig.set_tight_layout(True)
ax1 = fig.add_subplot(211)
ax2 = fig.add_subplot(212)
ax1.grid()
ax2.grid()

for qt in risk_record:
    bt = risk_record[qt]['bt']

    data = bt[[u'tradeDate',u'portfolio_value',u'benchmark_return']]
    data['portfolio_return'] = data.portfolio_value/data.portfolio_value.shift(1) - 1.0   # 总头寸每日回报率
    data['portfolio_return'].ix[0] = data['portfolio_value'].ix[0]/	10000000.0 - 1.0
    data['excess_return'] = data.portfolio_return - data.benchmark_return                 # 总头寸每日超额回报率
    data['excess'] = data.excess_return + 1.0
    data['excess'] = data.excess.cumprod()                # 总头寸对冲指数后的净值序列
    data['portfolio'] = data.portfolio_return + 1.0     
    data['portfolio'] = data.portfolio.cumprod()          # 总头寸不对冲时的净值序列
    data['benchmark'] = data.benchmark_return + 1.0
    data['benchmark'] = data.benchmark.cumprod()          # benchmark的净值序列
    risk_record[qt]['hedged_max_drawdown'] = max([1 - v/max(1, max(data['excess'][:i+1])) for i,v in enumerate(data['excess'])])  # 对冲后净值最大回撤
    risk_record[qt]['hedged_volatility'] = np.std(data['excess_return'])*np.sqrt(252)
    risk_record[qt]['hedged_annualized_return'] = (data['excess'].values[-1])**(252.0/len(data['excess'])) - 1.0
    # data[['portfolio','benchmark','excess']].plot(figsize=(12,8))
    # ax.plot(data[['portfolio','benchmark','excess']], label=str(qt))
    ax1.plot(data['tradeDate'], data[['portfolio']], label=str(qt))
    ax2.plot(data['tradeDate'], data[['excess']], label=str(qt))
    

ax1.legend(loc=0)
ax2.legend(loc=0)
ax1.set_ylabel(u"净值", fontproperties=font, fontsize=16)
ax2.set_ylabel(u"对冲净值", fontproperties=font, fontsize=16)
ax1.set_title(u"PB - 不同参数分组选股净值走势", fontproperties=font, fontsize=16)
ax2.set_title(u"PB - 不同参数分组选股对冲沪深300指数后净值走势", fontproperties=font, fontsize=16)
```

这种结果在情理之中，原因是PB低的往往是大盘蓝筹股，而PB较高的是中小创。接下来中性化后（去除行业因素），看看结果会如何？
