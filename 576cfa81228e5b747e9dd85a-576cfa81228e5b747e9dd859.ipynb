**Alpha#02**
$$(-1 * correlation(rank(delta(log(volume), 2)), rank(((close - open) / open)), 6))$$

先了解下两个运算符：<br>
**Rank(x)** ： 排列所有工具的 x 值，然后返回的数值是介于0.0和1.0之间均匀分布的浮点数。<br>
例如假设你有一个向量：<br>
A = [29, 12, 5, 7, 10]<br>
现在，将会为向量里的每个元素**按升序分级**，最小一级为 0：<br>
A = [29, 12, 5, 7, 10]<br>
A' = [4, 3, 0, 1, 2] <br>
每个“等级”再除以（元素个数 - 1）。于是得到：<br>
Rank (A) = [1.00, 0.75, 0.00, 0.25, 0.50]<br>
所以您等到了一个从 0 到 1 间距相等的向量。<br>

**Correlation(x, y, n)** : 	计算向量 x 中的值在过去n天的相关系数。请注意，n 必须小于256。

先看公式，主要涉及了volume、close、open变量，初看起来又是从量价关系为出发点的策略，之前写的[Alpha#03](https://uqer.datayes.com/community/share/5750f4a1228e5b86a61fd7dc),就是基于量价背离的思想的。再仔细分析公式，主要有以下几点：

1.计算了成交量的对数，并计算了昨日与前日的对数差。<br>
这刻画了成交量的日变化量，由于对数化，其值越小表示成交量变化不大，成交量趋于平缓；其值越大表明有相对的增量或缩量。

2.(close - open)/open，计算了当日的收益率。公式计算了日收益与对数成交量的差分，并给负相关的股票给予更高权重。我们不妨分类讨论下：，<br>
假如成交量的差变小，说明成交量变化趋于平稳，此时给予日收益率大的股票更高权重。此时买的股票多为上涨且成交量平稳的，可以看作是趋势未变，持有者可以安心持股．未持者可以适当买入；<br>

假如成交量的差变大，说明成交量变化变大，出现了相对放量或缩量，此时给予昨日收益率小的股票更高权重。<br>
(1)相对放量，买入昨天跌的股票，因为量为价先，后续反弹的可能性较高。但也要根据股价目前位置而定：若在高位放量滞涨或下跌，高位放量滞涨就是股价经过长期炒作后已经处于相当的高度（或许已经翻了数倍），然后在一段较短时期内出现成交量不断放大而价格却停滞不前的情况（当然也可能创出了历史新高，但总之涨幅较小），此时庄家出货概率较大，后市不容乐观

(2)相对缩量，买入昨天跌的股票，表示缩量下跌，原因可能是：庄家为了吸货，打压股价，但是股票成交量不大，说明散户惜售，这时应该买入持股，庄即将拉升股票，以便吸货； 市场进入熊市，阴跌不止。
