# 格雷厄姆神奇公式真的有用吗？

# 策略思路(参考格林布拉特的《股市稳赚》):
- 首先，根据资本收益率ROA，将股票池中的所有上市公司进行排序，资本收益率最高的公司排在第1位，资本收益率最低的公司排在最后一位；
- 其次，同理，按股票收益率按从高到底进行排序；
- 最后，将这两个排名综合在一起。例如，一家资本收益率排名第23位，股票收益率排名第15位的公司，它的综合排名是第38位（23+15）。如果一家公司资本收益率排名第1位，而股票收益率排名第200位，那它的排名将是第201位。
---
神奇公式不是要找到资本收益率最高或者股票收益率最高的公司，而是要找到将这两个因素进行最佳组合的公司。

    从回测结果看，格雷厄姆的神奇公式不适合做短期投资，因为收益极为不稳，有时候都没有跑过大盘，只有长期（6~10年）才能显现出其稳健的收益率水平。
