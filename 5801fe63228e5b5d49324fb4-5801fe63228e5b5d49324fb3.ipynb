```python
import pandas as pd
from CAL.PyCAL import *
```

```python
cal = Calendar('China.SSE')
bgn = Date.todaysDate() - '1M'
end = Date.todaysDate()
preWorkingDay =cal.bizDatesList(bgn,end)[-1]
preWorkingDay.toISO()
```

```python
HK2RMB = DataAPI.MktFxRefRateGet(currencyPair=u"HKD/CNY",beginDate=u"",endDate=u"",field=u"tradeDate,midRate",pandas="1")['midRate'].iloc[0]
```

```python
df = DataAPI.SecIDGet(partyID=u"",assetClass=u"E",ticker=u"",cnSpell=u"",field=u"",pandas="1")
df = df[df['listStatusCD']!='DE']
sec_CN = df[df['exchangeCD'].str.contains('XSH')]
sec_HK = df[df['exchangeCD'].str.contains('XHKG')]
AH = pd.merge(sec_CN,sec_HK, on = 'secShortName')
xchg_HK = DataAPI.MktHKEqudGet(secID=AH['secID_y'].values.tolist(),ticker=u"",tradeDate=preWorkingDay.toISO(),beginDate=u"",endDate=u"",field=u"",pandas="1")
xchg_CN = DataAPI.MktEqudLatelyGet(secID=AH['secID_x'].values.tolist(),ticker="",field="",pandas="1")
xchg_CN['Chg'] = (xchg_CN['closePrice'] - xchg_CN['actPreClosePrice']) / xchg_CN['actPreClosePrice']
xchg_HK['Chg'] = (xchg_HK['closePrice'] - xchg_HK['actPreClosePrice']) / xchg_HK['actPreClosePrice']
xchg_CN = xchg_CN[['ticker','secShortName','closePrice','Chg']]
xchg_HK = xchg_HK[['ticker','secShortName','closePrice','Chg']]
```

```python
res = pd.merge(xchg_CN,xchg_HK,on='secShortName',suffixes=('_CN','_HK'))
res['closePrice_HKDinRMB'] = res['closePrice_HK'] * HK2RMB
res['H2A'] = res['closePrice_HKDinRMB'] / res['closePrice_CN']
#福耀玻璃	600660	3606	17.690	-1.060%	23.100	19.998	0.000%	113.05%
res = res[['secShortName','ticker_CN','ticker_HK','closePrice_CN','Chg_CN','closePrice_HK','closePrice_HKDinRMB','Chg_HK','H2A']]

def beautify_res(res):
    res['Chg_CN'] = res['Chg_CN'].apply("{0:.2%}".format)
    res['Chg_HK'] = res['Chg_HK'].apply("{0:.2%}".format)
    res['H2A'] = res['H2A'].apply("{0:.2%}".format)
    res = res.rename(columns={'secShortName':'中文简称',
                        'ticker_CN':'A股代码',
                        'ticker_HK':'H股代码',
                        'closePrice_CN':'A股价格',
                        'Chg_CN':'A股涨跌幅',
                        'closePrice_HK':'H股价格(港币)',
                        'closePrice_HKDinRMB':'H股价格(人民币)',
                        'Chg_HK':'港股涨跌幅',
                        'H2A':'比价:(H/A)'
                       })
    return res
#------------------------------
#put filter code here
res = res.sort_values('H2A',ascending=False)
#------------------------------
beautify_res(res)
```
