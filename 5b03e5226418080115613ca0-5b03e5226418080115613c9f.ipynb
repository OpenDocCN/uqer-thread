# 中国股票价格尾数的研究
    股票价格的尾数在某些数字上更多的出现， 我们称之为价格聚类现象。1962年，Osborne首先发现了美国股票市场上有关价格尾数 “0” 和 “5” 出现频率更高的价格聚类现象。本帖将探究这一现象是否出现在中国股票价格尾数中，并做实证研究。
    除此之外，在传统文化里，人民比较偏爱吉利数字 “8”，而回避不吉利的数字 “4”。这一中国传统文化现象是否对股票价格尾数产生影响，我们也实证研究一番。

## 1 考察日收盘价尾数特征
get_last_digit_from_num() 是获取价格尾数的函数。

```python
#!/usr/bin/python
#-*-coding:utf-8-*-

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math

# 获取价格的尾数，股票价格默认两位小数
# 小于两位小数，尾数取0；大于等于两位小数，取尾数。
def get_last_digit_from_num(price):
    s = str(price)
    if s.isdigit():
        return 0
    elif len(s.split(".")[1]) <= 1:
        return 0
    else:
        return int(s.split(".")[1][-1])
```

## 1.6 总结
>* 股票价格尾数出现了 “5” 的聚类现象，但是 “0” 概率分布除了近一年来是出现概率最高，其它时间段都出现反常，出现概率最低。如果排除数据源上面的问题，要么是中国股市本来存在 股价尾数出现 “0” 概率低的现象，要么是收盘价不能很好的描绘股票价格尾数分布情况，其它数据，如分时、分笔成交数据和买卖盘数据可能更好一点，更连续点。
>* 中国传统文化对收盘价尾数 “4” 和 “8” 的概率分布还是有一定影响的。吉利数 “8” 出现概率要大于不吉利数字 “4” 出现概率。
