## 【首席执行官的风采】

- 第一天上午，由通联数据的首席执行官王政博士为大家做培训。王政博士具有20年证券从业经验。1995年至1997年在哈佛大学从事博士后研究工作。1997年8月起先后在美国的道琼斯公司、彭博公司、巴克莱全球投资公司工作。2009年5月加入博时基金管理有限公司。现任股票投资部总经理、ETF及量化投资组投资总监兼博时上证超大盘ETF及联接基金基金经理、博时大中华亚太精选股票基金基金经理、博时深证基本面200ETF及联接基金经理。曾经的公募大佬，国内量化届的元老，这个量级的培训也是让小弟大开眼界。

- 王政博士的演讲立足于从大方向讲解量化投资的发展之路。首先介绍了传统投资的策略分析方法，包括基本面投资、技术面分析、主题投资和事件投资，提到了经典而合理的估值方法，如DCF模型、RIM模型、比率分析等等，并详细解释了股票的定价理论。在王总看来，量化投资应该是一个两级多维的模型，上层负责投资组合构建和统筹各个功能模块，下层的功能模块需包括收益预测、风险管理、费用管理、归因分析等等，前面两者用于控制风险并寻找收益，费用管理用于交易优化，而归因分析用于投资绩效评估和反馈。几者结合在一起便构成了相对完整的量化投研平台。王总还给大家详细介绍了他对于大数据和机器学习在量化中应用的看法，machine learning是未来高效化投资的趋势，无论从效率还是从信息挖掘的合理性，它都有人无法比拟的优势。通联自己也将deep learning用在了自己的研究和策略开发当中。听完王总的演讲，作为技术派拥护者的小弟兴奋的流下了鼻血，仿佛又踌躇满志的走在了庄康大道上（反正吹逼不要钱）。

## 【私募牛人的实盘经验分享】
- 当天下午，由上海桐昇通惠资产管理有限公司的投资总监张鹏程为大家上了精彩的一课。张总兼具北京大学数学学士和清华大学统计学硕士的背景，毕业后首先任职博时基金管理有限公司，负责社保基金第一只指数增强产品的管理，由于业绩突出获得社保基金大幅增资。后到桐昇通惠创业，任职投资总监和基金经理，有多年的量化投资和研究经验。

- 都说自古私募量化出牛人，这说法也是有科学依据的。从张总的演讲当中，我看到了一个职业做量化投资研究人的严谨和对事业的热爱。对于投资中的每个细节，他都会慎重的去考虑和研究，从其对于数据准确性的高要求便可以看出。张总分享了很多自己在因子投资研究的经验，其对股票收益率来源的解析小弟非常认同，与自己之前听过bloomberg的路演有异曲同工之妙。我结合自己听的报告讲解一下自己对于股票收益率的看法，股票收益率可分解为：来自①市场$\beta$的部分、②来自行业配置的部分、③来自风格暴露的部分和④超额收益$\alpha$。其中，②和③是来自$Smart\  \beta$的收益，他们本身应该归因在风险因子里面，只不过这些风险因子也会带来超额收益。在量化研究中，归因分析也是很重要的一点，它能让你有效的找到赚钱的因素具体来自于哪儿，使我们可以用最简单有效的Reinforce learning的理论去不断优化对于模型的配置。下午张总的演讲验证了我对于这股票收益率的认知，也是感觉帮帮哒。
