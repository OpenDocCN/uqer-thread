## 一、即期利率模型的波动率结构

以最简单的一元随机偏微分方程模型为例，一般而言short rate（利率无限可分）模型都有固定的$\beta$, 例如[Ho & Lee](https://www.jstor.org/stable/2328161)($\beta=0$), [Vasicek](https://www.sciencedirect.com/science/article/pii/0304405X77900162)($\beta=0$), [Cox, Ingersoll & Ross](http://pages.stern.nyu.edu/~dbackus/BCZ/discrete_time/CIR_Econometrica_85.pdf)($\beta=1/2$)。如若$\beta=1$，即期利率便有了对数正态的波动率结构。
$$ dr = u(r)dt + vr^{\beta}dW $$
我们将即期利率波动率结构定义为:
$$\omega(r)=vr^{\beta}$$
需要注意的是这里参数$u(r)与\omega(r)$不依赖时间$t$。
我们可以从SHIBOR数据中得到每一期$\delta r_i$, 如果数据频率是每天，那么$\delta t=1/252$，对上述short rate SDE求期望，得
$$E[(\delta r)^2]=\omega(r)^2\delta r=(vr^{\beta})^2\delta t$$

两边取对数
$$ln(E[(\delta r)^2])=ln((vr^{\beta})^2\delta t)=2\beta*ln(r)+ln(v^2)+ln(\delta t)$$

至此，我们可以构建线性回归：
$$ln((\delta r_i)^2)=\beta_0+\beta_1*ln(r_i)+\epsilon_i$$

用Ordinary least squares(OLS)估计参数$\hat{\beta_0}$与$\hat{\beta_1}$

$$\hat{\beta_0}=ln(v^2)+ln(\delta t);\hat{\beta_1}=2\beta$$

最终，确定$v, \beta$的具体数值
$$v=e^{\frac{1}{2} (\hat{\beta_0}-ln(\delta t))};\beta=\frac{1}{2}\hat{\beta_1}$$

```python
# 从API获取的是一个pandas.core.frame.DataFrame
Shibor1M = DataAPI.MktIborGet(secID=u"",ticker=u"Shibor1M",tradeDate=u"",beginDate=u"",endDate=u"",currency=u"",field=u"",pandas="1")
Shibor3M = DataAPI.MktIborGet(secID=u"",ticker=u"Shibor3M",tradeDate=u"",beginDate=u"",endDate=u"",currency=u"",field=u"",pandas="1")
# 将日期设为index
Shibor1M.index = Shibor1M['tradeDate']
Shibor1M = Shibor1M['rate']
Shibor3M.index = Shibor3M['tradeDate']
Shibor3M = Shibor3M['rate']
Shibor1M.head()
```

```python
# Required Packages
import numpy as np
import pandas as pd
# 百分化
data = Shibor1M
# r 与 dr^2
r = data[1:]
dr2 = data.diff()**2
dr2 = dr2[1:]
# 取对数
lndr2 = -np.log(dr2)
lnr = -np.log(r)
# 组成一个用于回归的Dataframe
regData = pd.DataFrame()
regData['lnr'] = lnr
regData['lndr2'] = lndr2
# 去除无穷与NaN
regData = regData.replace([np.inf, -np.inf], np.nan).dropna(axis=0, how="any")
# 输出index
date = regData.index
regData.head()
```

```python
# Required Packages
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

# Linear Regression， smf近似R语言，自带截距
result = smf.ols('lndr2 ~ lnr', data=regData).fit()
print(result.summary())

# 得到v与beta最终数值
# 此处截距取负值是由于先前取log后进行回归的值为正
beta0 = -result.params[0];
beta1 = result.params[1];
v = np.exp(0.5*(beta0-np.log(1/252.0)));
beta = 0.5*beta1;
print(v, beta)

# 绘图
plt.scatter(regData['lnr'], regData['lndr2'], marker = 'o', label="data")
plt.plot(regData['lnr'], result.fittedvalues, c= 'r', label="OLS")
plt.legend(loc='best')
```

## 二、基于稳态概率密度定义漂移项

### 1.Fokker–Planck equation
`这里试了很多方法，似乎不能实现长公式换行。`
$$
\frac{\partial}{\partial t}p(r,t) = -\frac{\partial}{\partial r}(u(r)p(r,t))+\frac{\partial^2}{\partial r^2}\left(\frac{1}{2} \omega(r)^2 p(r,t)\right) = -\frac{\partial}{\partial r}(u(r)p(r,t))+\frac{\partial^2}{\partial r^2}\left(\frac{1}{2} (vr^\beta)^2 p(r,t)\right)
$$

$p(r,t)$为即期利率$r$的概率密度函数

至此，我们有
$$\frac{\partial p}{\partial t} = \frac{1}{2} v^2 \frac{\partial^2}{\partial r^2}(r^{2\beta} p) -(u(r)p)$$

当$p(r,t) \rightarrow p_\infty(r)$时，
$$
\frac{\partial p}{\partial t} = \frac{1}{2} v^2 \frac{\partial^2}{\partial r^2}(r^{2\beta}p_\infty) - \frac{\partial}{\partial r}(u(r)p_\infty)=0
$$

即
$$
\frac{\partial}{\partial r}(u(r)p_\infty) = \frac{1}{2} v^2 \frac{\partial^2}{\partial r^2}(r^{2\beta}p_\infty)
$$

两边积分
$$ u(r)p_\infty = \frac{1}{2} v^2 \int\frac{\partial^2}{\partial r^2}(r^{2\beta}p_\infty) dr$$
$$ u(r)p_\infty = v^2 \beta r^{2\beta-1} p_\infty + \frac{1}{2}  v^2 r^{2\beta} \frac{\partial}{\partial r}(ln(p_\infty))$$

### 2.经验分布的拟合
为获得$p_\infty$，我们需要指定一个分布来拟合即期利率$r$的经验分布。
原文中Ahmad and Wilmott (2006)选用对数正态分布。
$$
f(x;\mu,\sigma)=\frac{1}{x\sigma \sqrt{2\pi}} exp\left(-\frac{(ln(x)-\mu)^2}{2 \sigma^2}\right) \rightarrow  p_\infty(r)=\frac{1}{r\sigma \sqrt{2\pi}} exp\left(-\frac{(ln(r)-\mu)^2}{2 \sigma^2}\right)
$$

将稳态概率密度$p_\infty(r)$带入$u(r)$

$$
u(r) = v^2 \beta r^{2\beta-1} + \frac{1}{2}  v^2 r^{2\beta} \frac{\partial}{\partial r}(ln(p_\infty)) = v^2 \beta r^{2\beta-1} \left(\beta-\frac{1}{2}(\frac{1}{\sigma \sqrt{2\pi}}+\frac{ln(r)-\mu}{\sigma^2})\right)
$$

用`scipy.stats`中的连续分布拟合参数$\hat{\mu}$,$\hat{\sigma}$, 得到漂移项

$$
\hat{u}(r) = v^2 \beta r^{2\beta-1} \left(\beta-\frac{1}{2}(\frac{1}{\hat{\sigma} \sqrt{2\pi}}+\frac{ln(r)-\hat{\mu}}{\hat{\sigma}^2})\right)
$$

```python
import matplotlib.pyplot as plt
import scipy
import scipy.stats
import seaborn as sns
sns.set()

# 数据
data2 = data * 100
y = data2.values
x = np.arange(min(y), max(y),  0.001)

# 对数正态分布
dist = scipy.stats.lognorm
param = dist.fit(y, floc=0)

# 正态分布
dist2 = scipy.stats.norm
param2 = dist2.fit(y)
print(param)
print(param2)

# 参数
mu = np.log(param[2])
sigma = param[0]
print(mu,sigma)

# 绘图
ax = data2.hist(bins = 20)
lognorm_fitted = dist.pdf(x, *param[:-2], loc=param[-2], scale=param[-1])
norm_fitted = dist2.pdf(x, *param2[:-2], loc=param2[-2], scale=param2[-1])
ax2 = ax.twinx()
ax2.plot(x, lognorm_fitted, label='lognormal')
ax2.plot(x, norm_fitted, label='normal')
plt.legend(loc='upper right')
plt.grid(False)
plt.show()
```

## 三、利用收益率曲线斜率求出利率风险市场价格

### 1.零息债券价格的近似
对于一个零息债券$Z(r,t;T)$
$$\frac{\partial Z}{\partial t} + \frac{1}{2} \omega^2 \frac{\partial^2 Z}{\partial r^2}(u-\lambda\omega) \frac{\partial Z}{\partial r} -rZ = 0 $$

在$t\rightarrow T$处Taylor展开，
$$Z(r,t;T)\sim Z(r,T;T)+a(r)(T-t)+b(r)(T-t)^2+\cdots$$
$$Z(r,T;T)=1$$
我们需要求出系数$a(r)$，$b(r)$，
$$\frac{\partial Z}{\partial t} = -a(r)-2b(r)(T-t) $$
$$\frac{\partial Z}{\partial r} = \frac{da}{dr}(T-t)+\frac{db}{dr}(T-t)^2 $$
$$\frac{\partial^2 Z}{\partial r^2} = \frac{d^2a}{dr^2}(T-t)+\frac{d^2b}{dr^2}(T-t)^2 $$

将以上偏微分带入零息债定价等式，
我们会得到
$$a(r)=-r$$
$$b(r)=\frac{1}{2}(r^2-(u-\lambda \omega))$$

带入之前的泰勒展开式，
$$Z(r,t;T)\sim Z(r,T;T)-r(T-t)+\frac{1}{2}(r^2-(u-\lambda \omega))(T-t)^2+\cdots$$

### 2.收益率曲线与利率风险市场价格的关系
对泰勒展开式两边取对数
$$ln(Z(r,t;T))=ln((Z-1)+1)=(Z-1)-\frac{(Z-1)^2}{2}+\cdots$$

我们忽略$(T-t)^3$及以上高次项，最终得到
$$-\frac{ln(Z)}{T-t}\sim r+\frac{1}{2}(u-\lambda\omega)(T-t)$$
$$r_L = -\frac{ln(Z)}{T-t}\approx r+\frac{1}{2}(u-\lambda\omega)(T-t)$$

在真实市场里，我们没有无限可分的可交易的利率，所以以一个较长期的利率$r_l$和一个较短期利率$r_s$代入上式，
$$r_l = r_s+\frac{1}{2}(u-\lambda\omega)(T_l-T_s)$$

计算收益率曲线的斜率
$$slope = \frac{1}{2}(u-\lambda\omega) = \frac{r_l-r_s}{T_l-T_s}$$

最终通过上述代入较短期较长期利率的等式，求得利率风险的市场价格，
$$\lambda(r) 
= \frac{1}{\omega}\left(\hat{u(r)}-2*\frac{r_l-r_s}{T_l-T_s} \right) $$

$$
= \frac{1}{vr^{\beta}} \left(v^2 \beta r^{2\beta-1} \left(\beta-\frac{1}{2}\left(\frac{1}{\hat{\sigma} \sqrt{2\pi}}+\frac{ln(r)-\hat{\mu}}{\hat{\sigma}^2}\right)\right)
-2*\frac{r_l-r_s}{T_l-T_s}\right)
$$

```python
import math
short = Shibor1M; # 1 Month SHIBOR rate
long = Shibor3M; # 3 Month SHIBOR rate
slope = (long-short)/(3/12.0-1/12.0);
u = np.multiply((v**2)*(short**(2*beta-1)),(beta-0.5*(1/(sigma*np.sqrt(2*math.pi))+(np.log(short)-mu)/(sigma**2))));
w = v*(short**beta);
lam = (1/w)*(u-2*slope);
lam.plot()
```

上图为以时间为基准的利率风险的市场价格，虽然我们之前的假设$\lambda$是即期利率$r$的函数，但以时间为轴却更能呈现出其特点。
