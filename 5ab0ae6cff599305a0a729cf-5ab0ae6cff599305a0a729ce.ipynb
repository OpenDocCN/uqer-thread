# 定义换月效应

换月效应是日历日效应的一种，表现为月末最后交易日前后资产收益率规律性的涨跌。

```python
def prepare_dataframe(sec_id):
    df300 = DataAPI.MktIdxdGet(indexID=sec_id,ticker=u"",beginDate=u"20051101",endDate=u"20171231",exchangeCD=u"",field=u"",pandas="1")
    df300.set_index('tradeDate',inplace=True)
    eom_idx = df300.index.get_indexer(calendar_df[calendar_df["isMonthEnd"] == 1]['calendarDate'])
    # for eom before the available history of asset, get_indexer returns -1
    eom_idx = np.trim_zeros(eom_idx+1)-1

    eom = df300.index[eom_idx]
    df300.reset_index(inplace=True)
    relative_to_eom_idx = {}
    for i in range(-10,11):
        # out of range index

        if i+eom_idx[-1] > df300.shape[0] - 1:
            relative_to_eom_idx.update({i:(i+eom_idx[:-1],eom[:-1].copy())})
        elif i+eom_idx[-1] < 0 :
            relative_to_eom_idx.update({i:(i+eom_idx[1:],eom[1:].copy())})
        else:
            relative_to_eom_idx.update({i:(i+eom_idx,eom.copy())})

    days_to_eom = []
    for dte, v in relative_to_eom_idx.items():
        df1 = pd.DataFrame(data={"days_to_eom":dte, 'eom':v[1]})
        df1.index = v[0]
        days_to_eom.append(df1)
    days_to_eom = pd.concat(days_to_eom, axis=0, verify_integrity=False)

    df300 = df300.join(days_to_eom,how = 'inner')

    df300["month"]=df300.apply(lambda x: int(x[u'tradeDate'][5:7]), axis=1)
    df300["year"] =df300.apply(lambda x: int(x[u'tradeDate'][:4]), axis=1)
    df300['eom_year'] = df300['eom'].apply(lambda x:int(x[:4]))
    df300['eom_month'] = df300['eom'].apply(lambda x:int(x[5:7]))
    def TOTM_stage(days_to_eom):
        if days_to_eom in (-1,-2,-3,-4):
            return 'pre4'
        elif days_to_eom in (0,1,2,3):
            return 'post4'
        else:
            return 'other'

    df300['TOTM_stage'] = df300.apply(lambda x: TOTM_stage(x['days_to_eom']) , axis=1)

    ## returns by days to eom

    ret_by_dte = df300.groupby('days_to_eom')['CHGPct'].mean()

    # avg daily return by month and stage
    ret_by_eom_n_stage = df300.groupby(['eom', 'TOTM_stage'])['CHGPct'].mean()
    ret_by_eom_n_stage.name = "CHGPct"
    ret_by_eom_n_stage = ret_by_eom_n_stage.reset_index()
    ret_by_eom_n_stage = ret_by_eom_n_stage.pivot(index='eom', columns='TOTM_stage', values='CHGPct')
    # daily return by eom
    ret_by_stage = ret_by_eom_n_stage.mean()

    # average daily return
    ret_by_eom = df300.groupby(['eom'])['CHGPct'].mean()
    ret_by_eom.name = 'daily_ret_by_eom'
    ret_by_eom_n_stage = ret_by_eom_n_stage.join(ret_by_eom)

    ret_by_eom_n_stage.reset_index(inplace=True)
    ret_by_eom_n_stage['eom_year'] = ret_by_eom_n_stage['eom'].apply(lambda x:int(x[:4]))
    ret_by_eom_n_stage['eom_month'] = ret_by_eom_n_stage['eom'].apply(lambda x:int(x[5:7]))

    # TOTM return by calendar month and year trend
    ret_by_eom_n_stage.groupby('eom_month').mean()
    ret_by_eom_n_stage.groupby('eom_year').mean()
    return ret_by_eom_n_stage, df300
```

```python
ret_by_eom_n_stage, df300 = prepare_dataframe(u"H00300.ZICN")
```

# 换月效应形成原因的探讨

## 基金经理的window dressing行为

公募基金在季报，半年报，年报中要披露持仓。基金经理的window dressing行为是指基金经理粉饰业绩的行为，在报告期末卖出组合中近期表现不佳的股票。在业绩考核的压力下，原来手持爱票扛了一个季度的基金经理终于决定忍痛割爱，产生抛压。这可能也是在熊市中换月效应更明显的原因。

## 周期性的月末流动性紧张

window dressing行为充其量只能解释季末的换月效应，而由商业银行造成的周期性流动性紧张则能更好解释全年的换月效应。商业银行在很长时间内都有存贷比考核指标，简单说就是监管要求放贷出去的钱占吸收存款的比例不能过高。在月末考核的时候，存贷比过高的银行为了降低比例会高息揽储或者推迟贷款发放。这些行为导致市场流动性紧张，提升了短期无风险收益率，由此促使一些追求短期收益的资金从股市撤出，导致了月末前的下跌。而月末考核结束后，由于市场因流动性的恢复而企稳。
2015年10月起监管取消了存贷比的强制限制，仅仅作为一个观察指标。我们观察到近两年（2016-2017）换月效应已经变得很弱了，这也暗含了两者之间或许存在因果关系。
