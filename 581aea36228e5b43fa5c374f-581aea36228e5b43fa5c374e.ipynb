一：DTW算法简介：
1：算法的目的在于：衡量两个长度不同的时间序列的相似度的方法。在时间序列中，需要比较相似性的两段时间序列的长度可能并不相等，在语音识别领域表现为不同人的语速不同（匹配串和被匹配串的长度不同）。那么此时传统的欧几里得距离无法有效地求的两个时间序列之间的距离（或者相似性）。
2：我们感兴趣的是波动，相同的波形趋势在我们看来很有可能是同样的语音。
3：大部分情况下，两个序列整体上具有非常相似的形状，但是这些形状在x轴上并不是对齐的。所以我们在比较他们的相似度之前，需要将其中一个（或者两个）序列在时间轴下warping扭曲，以达到更好的对齐。而DTW就是实现这种warping扭曲的一种有效方法。DTW通过把时间序列进行延伸和缩短（或者是元素点的近邻匹配），来计算两个时间序列性之间的相似性。
4：DTW是一个典型的优化问题，它用满足一定条件的的时间规整函数W(n)描述测试匹配和被匹配串的时间对应关系，求解两个串匹配时累计距离最小所对应的规整函数。

二：算法
1：算法的背景：
	（1）：P Q两个串，分别是m维,n维。构造一个m*n的矩阵（ij元素为dist（P[i]-Q[j]）），（给了PQ就可以直接获得的矩阵）
    我们要做的就是在这个矩阵中找到一个路径，从左下角走到右上角，使走过的路径上的数的和最小。（动态规划问题）
    （2）：方向性：每次只可以往上左右走，每次走一个格
    （3）：假设两个多变量的时间序列X={x_1,x_2,...,x_m} Y={y_1,...y_n} 其中每个x_i，y_i是q维的多变量样本，在我们规定了距离之后即d（x_i,y_i）既可以计算			 X,Y之间的动态距离Distance(X,Y)=D(m,n)
			(1): i=0 或 j=0时 D(i,j)=0
			(2)：i(1:m) j(1:n)循环计算 d_ij=d(x_i,y_j)
				D（i,j）=d_ij+min{D(i-1,j-1),D(i-1,j),D(i,j-1)}
				最终获得的D（m,n）就是最后X与Y的动态规划距离。
    

```python
import numpy as np
import pandas as pd
def dist(aa,bb): #内部的距离的定义
    return abs(aa-bb)
def trans(a):
    l=[]
    for i in range(1,len(a)+1):
        l.append(a[-i])
    return l
def dtw(t,r):
    n=len(t)
    m=len(r)
    d=np.zeros((n,m))
    t=trans(t)
    r=trans(r)
    for i in range(n):
        for j in range(m):
            d[i][j]=dist(t[i],r[j])
    realmax=10000000
    D=np.ones((n,m))*realmax
    D[0][0]=d[0][0]
    #动态规划
    #print d
    #print D
    for i in range(1,n):
        for j in range(0,m):
            D1=D[i-1][j]
            if j>0:
                D2=D[i-1][j-1]
            else:
                D2=realmax
            if j>1:
                D3=D[i-1][j-2]
            else:
                D3=realmax
            D[i][j]=d[i][j]+min(D1,D2,D3)
            #print D
    return D[n-1][m-1]
def DTW(a,b):
    m=len(a)
    n=len(b)
    dis=[]
    for i in range(1,m):
        dis.append(dtw(a[0:i],b))
    for j in range(1,n):
        dis.append(dtw(a,b[0:j]))
    return min(dis)
def normalized(a):#为何要用标准化
    m=np.mean(a)
    s=np.std(a)
    return (a-m)/s
```

实验二 上证指数的分析（仅仅用close这个一维数据）
(也可以进行个股的分析，只需在程序中改一下标注即可（见注释）)

```python
now="20161101"
day="20120101"
window_length=20
ticker_name='000001'
#数据载入
#name_type='preClosePrice'      #股票前收盘价的名字
name_type='closeIndex'          #指数收盘价的名字
#数据载入
#data=DataAPI.MktEqudGet(ticker=ticker_name,beginDate=day,endDate=now,field=[],pandas="1")  #获得了历史上的股票数据
data=DataAPI.MktIdxdGet(ticker=ticker_name,beginDate=day,endDate=now,field=u"",pandas="1")  #指数数据
#print data.head()
close=data[name_type].values
###########################################可以调用任何有用的数据研究
#print data
date=data['tradeDate'].values
#print date
distance=[]
begindate=[]
enddate=[]
x=close[-window_length:]#长度window_length的窗口 获得末尾数据
#数据标准化
t=normalized(x)               #这是待匹配的序列 （时间段中的末尾数据）
nowbegin=date[-window_length] #起始时间 指待匹配的序列的起始时间
nowend=date[-1]               #终止时间
T1=[]                         #算收益率的序列
T2=[]
for i in range(len(close)-window_length-4):#共有五个序列窗口在读数据
    begindate.append(date[i])
    enddate.append(date[i+window_length-1])#第一段 第二段 不断往后排
    T1.append((close[i+window_length]-close[i+window_length-1])/close[i+window_length-1])
    T2.append((close[i+window_length+1]-close[i+window_length])/close[i+window_length])
    r=normalized(close[i:i+window_length])#待匹配的样本历史每次循环都取出一段数据
    distance.append(DTW(t,r)) ###################距离认为是成交量和价格匹配的误差和 这个也可以改
#print "done"
w=[distance,T1,T2]
w=np.transpose(w)#变成矩阵列向量
c=pd.DataFrame(w,columns=['distance','T1','T2'])#转成dataframe
a=pd.DataFrame(begindate,columns=['begindate'])
b=pd.DataFrame(enddate,columns=['enddate'])
a=a.join(b,how='outer')#ab先合成一起
record=a.join(c,how='outer')
record=record.sort(columns='distance')
#print 'Begindate:  %s' %nowbegin
#print 'Enddate:    %s' %nowend
#print 'distance前五'
match_result=record[0:5]
```
