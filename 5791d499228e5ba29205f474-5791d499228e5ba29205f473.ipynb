**Alpha#37**
$$(rank(correlation(delay((open - close), 1), close, 200)) + rank((open - close)))$$

先了解下两个运算符：<br>
**Rank(x)** ： 排列所有工具的 x 值，然后返回的数值是介于0.0和1.0之间均匀分布的浮点数。<br>
例如假设你有一个向量：<br>
A = [29, 12, 5, 7, 10]<br>
现在，将会为向量里的每个元素**按升序分级**，最小一级为 0：<br>
A = [29, 12, 5, 7, 10]<br>
A' = [4, 3, 0, 1, 2] <br>
每个“等级”再除以（元素个数 - 1）。于是得到：<br>
Rank (A) = [1.00, 0.75, 0.00, 0.25, 0.50]<br>
所以您等到了一个从 0 到 1 间距相等的向量。<br>

**Correlation(x, y, n)** : 	计算向量 x 中的值在过去n天的相关系数。请注意，n 必须小于256。

**Delta(x, n)** :  	x[date] - x[date - n]. 请注意，n 必须小于256

首先，观察公式，我们发现主要用到的数据是open和close，看来主要是从“价”的角度去构建的alpha。

（1）看公式的左半部分，发现计算了delay((open-close),1)和close过去200天的相关系数。 open-close很好理解，就是开盘比收盘高的点数。open-close为正的话，说明当天的股价是**跌**的。由于我们的数据只能取道前一天的，因此**delay1，计算的是前第二天**的数据。

我们知道alpha公式计算的结果越大，我们会分给该股票更高的权重。因此，我们希望得到的结果是过去200天，当天收盘价与**前一天**“涨跌幅”**正相关**的！ 也就是，**前一天跌幅序列升高时，今天的收盘价序列也升高，前一天的跌幅序列降低时，收盘价序列也降低。**

也就是说，我们希望找到这类股票：**如果前一天股价跌/涨了，第二天股价上涨/下跌的概率大的**。

（2）显然，我们不希望前一天股价涨，今天股价下跌的情况。因此，再看公式右半部分，rank(open-close)。**刚好给与了前一天股价下跌最多这种情况更大的权重**，因此，公式表达出了偏好：选取前一天股价跌了，第二天股价上涨概率最大的

综合来看，公式的左半部分是找出前一天股价涨跌与第二天股价升降**相关性最高**的，右半部分则是在相关性高的股票中选择前一天股价**下跌**最多的。这里用到的是“均值回归”的思想。
