# 菜鸟的第一篇策略仿写

发现一个问题：
DataAPI.MktIdxFactorOneDayGet，其中tradeDate形式'20180102'不报错，'2018-01-02'报错

根据https://uqer.datayes.com/v3/community/share/575e54d2228e5b8a37285c79中的策略条件
1.大股票池“中证500”标的个股；
2.按市值排列，取市值最小的100只个股；
3.剔除停牌、t-1日涨跌停个股；（假如剩余m只个股）
4.选出这m只个股中5日均线大于10日均线的个股；（假如剩余n只）
5.判断大盘条件：t-2日的20日均线值>t-1日的20日均线值，则只卖不买；或者大盘t-1日收于5日均线以下则只卖不买；
6.如果非5条件，则选择市值最小的3只个股每只持仓30%；
7.每天重复1~6，如果t日筛选出的的个股于t-1日的个股出现相同的则，该股不动；
在判断大势的时候会筛选出3支票：a、b、c，假如大盘不好，但是已经持有a，f，x；那就只卖f、x，而保留a
