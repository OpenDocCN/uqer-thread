**02.call_相对市盈率因子选股**
地址：https://uqer.datayes.com/community/share/586dfe7089e3ba004eefe2e5
取股票的TTM市盈率数据(此处舍去市盈率为负和市盈率数值特别大的股票)，该数据再和每天的全市场市盈率均值求比值，该比值可以认为是该股票目前的市盈率在全市场中的排序位置；
对每只股票，第一步中求得的比值再求一段时间的均值(本文采用六个月)，该均值可认为是股票市盈率在全市场排序位置的参考值；
将第二步中的均值除以第一步中的比值，得到的值可以表征股票是否高估或低估，这个值大于1则股票低估，这个值小于1则股票高估，这个值就是相对市盈率因子

思路：
1,整体感觉上是股价会回归的思路（反转）
2,6个月可能过长了，3个月感觉会好一些，
3,不明白为何用5天的秩相关序数来判定和好坏，由于市盈率基本和股价同频变动，这次在今天在池子中的，明天可能也在，感觉至少是2周或者3月的样子较好，当然要结合第一项的x月平均，也要增加。

**07.call_非流动性因子ILLIQ - 刨根问底之中性化检验**
地址：https://uqer.datayes.com/community/share/57c6730b228e5b6d227c7314

一段时间内（本文N为5日），每日价格变化幅度绝对值和成交额的比值求平均，ILLIQ 越小流动性好；ILLIQ 越大，则说明流动性越差，所以ILLIQ 表示证券的非流动性。受益也越高。
思考：
1，市值，换手率暴露等无法避免的，随机选股都可能有暴露。何况有因子和板块，市值等有不可避免关联性。
2，可考虑新股的暴露性测试。（上市天数）
3，有地方采用排序作为分位过滤条件，大部分采用正态分布的切分点，是否不确切，未能判断数据分布为正态分布，用排序作为分位可能更有说服力。
4，代码不是很容易看懂哇
5，采用秩序数作为判断因子有效性比较合理，避免了采用绝对数据的参数选择问题。同时也对外部市场的强弱势免疫。
6，采用1日的信息，预测20日价格的是否不确切，偶然因素较大。比如今天高明天低，恰好选中今天，就入池了（整体概率上看也是中性的，也有部分是今天低，明天高，未入池，对称的）

2,人们往往会认为凯利公式所针对投注比例是全资产，但其实上并不,凯利公式所针对的投注比例是你可承受损失的资产。比如说，有一个私募，投入了1000w，但是有0.8的清仓线，其实只能承受200w的损失。那么在凯利公式里，f针对的本金其实只有200w，也就是说，如果f=0.7，你一次压上去的钱只有 200w*0.7=140w。
