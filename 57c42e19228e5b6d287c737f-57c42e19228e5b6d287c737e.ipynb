```python
def generateFactorOrderList(buyTarget, account, today, maxPercent = 0.2):
    orderList = pd.DataFrame(columns = ['secID', 'amount'])
    cashAmount = account.cash
    currentValidHolding = account.valid_secpos
    for stk in currentValidHolding:
        if stk not in buyTarget:
            referencePrice = account.referencePrice.get(stk, 0.0)
            if np.isnan(referencePrice) or referencePrice == 0.0:
                print "Fail to sell stock " + stk + " on date " + str(today)
            else:
                amount = 0 - currentValidHolding.get(stk, 0)
                orderList.loc[len(orderList)] = [stk, amount]
                cashAmount = cashAmount - amount * referencePrice
    validBuyTarget = []
    for stk in buyTarget:
        referencePrice = account.referencePrice.get(stk, 0.0)
        if np.isnan(referencePrice) or referencePrice == 0.0:
            print "Fail to buy stock " + stk + " on date " + str(today)
        else:
            validBuyTarget.append(stk)
            cashAmount = cashAmount + referencePrice * currentValidHolding.get(stk, 0)
    if len(validBuyTarget) > 0:
        buyAmount = min(cashAmount / float(len(validBuyTarget)), maxPercent * account.referencePortfolioValue)
        for stk in validBuyTarget:
            referencePrice = account.referencePrice[stk]
            amount = int(buyAmount / float(referencePrice) - currentValidHolding.get(stk, 0))
            orderList.loc[len(orderList)] = [stk, amount]
    orderList.sort(columns = 'amount', ascending = True, inplace = True)
    return orderList
```
