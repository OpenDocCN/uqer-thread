还有两点容易想到优化点
1，突破箱头和止损的判断可以是动态的

止损：高箱1/3头部，低箱以箱底为准，避免频繁操作
2，右移优化
为了避免未来函数，生成的信号右移了5单位，虽然100%避免了未来信息，但是也浪费了前几天的价格信息。
实际可以使用右移后的信号（5天前的信号）+转折日到今日的价格信息，同样没有使用未来信息(暂未处理)，


```python
#信号右移，等于手工延迟信号，消除未来函数
import numpy as np
import pandas as pd

stk_data=DataAPI.MktFunddGet(tradeDate=u"",secID=u"159919.XSHE",ticker=u"",beginDate=u"20150101",endDate=u"20170101",field=u"",pandas="1")

stk_data['MA5']=pd.rolling_mean(stk_data['closePrice'],window=5)
#提取极值点
max_price=[0 for i in range(0,len(stk_data))]
min_price=[0 for i in range(0,len(stk_data))]
extre_range=5
for i in xrange(extre_range,len(stk_data)):
    if stk_data['closePrice'][i] == stk_data['closePrice'][i-extre_range:i+1].max():
        max_price[i]=1
        max_price[i-extre_range:i]=[0 for ii in range(0,extre_range)]
    elif  stk_data['closePrice'][i] == stk_data['closePrice'][i-extre_range:i+1].min():
        min_price[i]=1
        min_price[i-extre_range:i]=[0 for ii in range(0,extre_range)]
stk_data['high_price']=stk_data['closePrice']*max_price
stk_data['low_price']=stk_data['closePrice']*min_price
stk_data['high_price']=(stk_data['high_price'].apply(lambda x:np.nan if x<0.01 else x)).fillna(method='pad')
stk_data['low_price']=(stk_data['low_price'].apply(lambda x:np.nan if x<0.01 else x)).fillna(method='pad')
#右移信号消除未来函数影响
stk_data['high_price']=stk_data['high_price'].shift(extre_range)
stk_data['low_price']=stk_data['low_price'].shift(extre_range)
stk_data[['closePrice','high_price','low_price']].plot(figsize=(14,8))


#生成买卖信号
stk_data['buy_sell']=[0 for li in range(0,len(stk_data))]
stk_data['stop_loss']=[0 for li in range(0,len(stk_data))]
high=0
low=0
for fi in range(extre_range,len(stk_data)):
    if np.isnan(stk_data['high_price'][fi]) or stk_data['high_price'][fi]<0.01 :
        continue
    high=stk_data['high_price'][fi]
    low=stk_data['low_price'][fi]
    #止损条件
    close_std=np.std(stk_data['closePrice'][fi-extre_range:fi+1])
    if (high-low)/low > 0.05:
        stopLoss=(2*high+low)/3
    else:
        stopLoss=low
    stk_data['stop_loss'][fi]=stopLoss
    #买入条件为高于high1.03
    if stk_data['closePrice'][fi]>high*1.0:#todo参数优化
        stk_data['buy_sell'][fi]=1
    elif stk_data['closePrice'][fi]<stopLoss:
        stk_data['buy_sell'][fi]=-1

#画图好看些
stk_data['buy_sell_show']=stk_data['buy_sell']*stk_data['closePrice'].mean()
stk_data[['high_price','low_price','closePrice','stop_loss']].plot(figsize=(16,8))    
```
