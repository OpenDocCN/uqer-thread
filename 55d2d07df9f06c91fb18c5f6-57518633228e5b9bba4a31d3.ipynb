3. 隐藏的未来数据
---------------------------------
***

> 金融理论中，常常提到的一个概念：过去价格的趋势对于未来价格趋势没有预测力。

3.1 基于沪深300指数周线的趋势测试
---------------------------------

```python
import datetime as dt
import numpy as np

data = DataAPI.MktIdxdGet(ticker='000300', beginDate='20060101', endDate='20150101')[['closeIndex', 'tradeDate']]
data['tradeDate'] = data['tradeDate'].apply(lambda x: dt.datetime(int(x[0:4]), int(x[5:7]), int(x[8:10])))
data = data.set_index('tradeDate')
data.head(5)
```

```python
weeklyLastPrice = data.resample(rule='w', how='last').dropna()
weeklyLastPrice.head()
```

```python
weeklyLastChange= np.log(weeklyLastPrice / weeklyLastPrice.shift(1))
weeklyLastChange.dropna(inplace=True)
weeklyLastChange.head()
```

```python
from statsmodels.graphics.tsaplots import plot_acf
from matplotlib import pyplot as plt

print(u"每周最后一个交易日价格收益自回归：{0:.4f}".format(weeklyLastChange['closeIndex'].autocorr()))

fig = plt.figure(figsize=(16,8))
ax = fig.add_subplot(111)
plot_acf(weeklyLastChange['closeIndex'], ax=ax)
```

看上去确实没有什么趋势性，理论似乎被证实了。

3.2 基于沪深300指数每周均价的趋势测试
-------------------------------------

```python
weeklyAveragePrice = data.resample(rule='w').dropna()
weeklyAveragePrice.head()
```

```python
weeklyAverageChange = np.log(weeklyAveragePrice / weeklyAveragePrice.shift(1))
weeklyAverageChange.dropna(inplace=True)
```

```python
weeklyAverageChange.tail()
```

```python

print("每周均价收益自回归: {0:.4f}".format(weeklyAverageChange['closeIndex'].autocorr()))
fig = plt.figure(figsize=(16,8))
ax = fig.add_subplot(111)
plot_acf(weeklyAverageChange['closeIndex'], ax=ax)
```

每周均价似乎有很大的趋势性，这个如何解释呢？
