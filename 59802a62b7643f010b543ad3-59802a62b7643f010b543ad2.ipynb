# MA均线缠绕的量化定义

最近写了一些关于股票的走势类型的选择代码，这里给大家献上MA均线缠绕的量化定义。
受API函数限制，此处只能一支股票一支股票的判断。
大家可以依据自己的标准修改参数，调整各均线间的距离关系，还可以在判断的两个时间间增加是否有当前一段时间的最大值或最小值发生，使该缠绕判断更准确。
最后，点赞、评论、收藏、关注的矿友股票都能涨停。

```python
#均线缠绕状态定义
import numpy as np
import pandas as pd
from datetime import datetime
from pandas import Series,DataFrame
from CAL.PyCAL import *
cal = Calendar('China.SSE')
eDate='2017-07-30'
bDate = cal.advanceDate(eDate,str(0-120)+'B',BizDayConvention.Preceding)
universe ='002561.XSHE'  # set_universe('A')[80:80]  #

def Get_MAwind(universe,bDate,eDate):
    if len(universe)<=0 :return []
    stdValue = 0.15   #std的最小值认为是缠绕状态
    WindPeriod = 10   #两次缠绕间最小间隔日期或一次缠绕的周期
    pf_MA=DataAPI.MktStockFactorsDateRangeGet(secID=universe,ticker=u"",beginDate=bDate,endDate=eDate,\
                                              field=u"tradeDate,MA5,MA10,MA20,MA60,MA120",pandas="1")
    pf_MA= pf_MA.set_index('tradeDate')
    pf_MA= pf_MA.dropna()
    pf_MA['std'] = pf_MA[['MA5','MA10','MA20','MA60','MA120']].std(axis=1)
    print pf_MA.sort(columns = 'std').head(10)
    if len(pf_MA[pf_MA['std']<=stdValue]) >= 2 and \
    (datetime.strptime(pf_MA[pf_MA['std'] <= stdValue].index.max(),"%Y-%m-%d") - datetime.strptime(pf_MA[pf_MA['std'] <= stdValue].index.min(),"%Y-%m-%d")).days > WindPeriod:
        print '存在缠绕'
    else:
        print '不符合缠绕定义'
    
Get_MAwind(universe,bDate,eDate)
```
