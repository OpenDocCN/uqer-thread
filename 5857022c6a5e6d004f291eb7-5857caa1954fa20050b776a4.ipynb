## A.动态买卖气指标 ADTM,STM,SBM
1.如果开盘价≤昨日开盘价，STM=0
如果开盘价>昨日开盘价，STM=(最高价-开盘价)和(开盘价-昨日开盘价)的较大值
2.如果开盘价≥昨日开盘价，DBM=0
如果开盘价<昨日开盘价，DBM=(开盘价-最低价)和(开盘价-昨日开盘价)的较大值
3.STM=DTM在N日内的和
4.SBM=DBM在N日内的和
5.如果STM>SBM,ADTM=(STM-SBM)/STM
如果STM<SBM,ADTM=(STM-SBM)/SBM
如果STM=SBM,ADTM=0
6.ADTMMA=ADTM的M日简单移动平均
7.参数N设置为23日，参数M设置为8日

** 简单交易示例 **
ADTM低于-.6代表太低，可买入。
ADTM上穿ADTMMA，且ADTM>0.1 时买入
ADTM下穿ADTMMA，清空

## B. 均幅指标
首先介绍真实波幅，是以下三个值中的最大值：

		1、当前交易日最高价和最低价的波幅
        2、前一交易日的收盘价与当前交易日最高价的波幅
		3、前一交易日的收盘价与当前交易日最低价的波幅

用公式写就是： $True Range = Max(High-Low,abs(High-PreClose),abs(PreClose-Low))$
$ATR14 = SMA(TrueRange，14)$
$ATR6 = SMA(TrueRange，6)$
![](http://wiki.mbalib.com/w/images/0/08/均幅指标.gif)

威尔德发现较高的ATR值常发生在市场底部，并伴随恐慌性抛盘。当其值较低时，则往往发生在合并以后的市场顶部

该指标可用于控制仓位，使得portfolio value的震荡控制在一定范围内。如[海龟交易](https://uqer.datayes.com/community/share/584682a8228e5b164c7700cf) 中。



## C. 乖离率 BIAS5,BIAS10,。。。
乖离率，也称Y值，是移动平均原理派生的一项技术指标，其功能主要是通过测算股价在波动过程中与移动平均线出现偏离的程度，从而得出股价在剧烈波动时因偏离移动平均趋势而造成可能的回档或反弹，以及股价在正常波动范围内移动而形成继续原有势的可信度。

如果股价偏离移动平均线太远，不管股价在移动平均线之上或之下，都有可能趋向平均线的这一条原理上。而乖离率则表示股价偏离趋向指标占百分比值。

$$Y=\frac{1}{0.01}\frac{p_t-SMA(p_t)}{SMA(p_t)}$$
一般乖离率 并不是很好，准确率不高，可参考使用下面的**E.CCI指标**
** 简单交易示例 **
Y>15, 代表大于平均的偏离值太大，清空；
Y<-10，代表小于平均的偏离值太大，买入；
-8<Y<-5，代表均值将下降，清空，
5<Y<12，代表均值将上升，买入

## D 布林带 BollUp, BollDown
1）计算MA

MA=N日内的收盘价之和÷N
2）计算标准差MD

MD=平方根N日的（C－MA）的两次方之和除以N
3）计算MB、UP、DN线

MB=（N－1）日的MA
UP=MB＋times[0]×MD
DN=MB－times[1]×MD
实际标准的时times=[2,2]，但是实现时不固定。

** 简单交易示例 **
卖出点：股价触碰布林带上轨（压力线） 或在均线下且下轨上
买入点：股价触碰布林带下轨 （支撑线）或在均线上且上轨下

## E 顺势指标 CCI5,CCI10,。。。
$$CCI=\frac{1}{0.015}\frac{p_t-SMA(p_t)}{\sigma(p_t)}$$

其中，$p_t$等于最高价、最低价和收盘价之和除以3.指标本身和乖离率很接近，分子相同。乖离率的分母是$\frac{1}{0.01\times SMA(p_t)}$, 这里是$\frac{1}{0.015\times\sigma (p_t)}$.
** 简单交易示例 **
设置股票池('HS300')
CCI5上穿其5日均线，且CCI10大于100，且CCI5>0, 处于上涨趋势的股票买入。
当CCI5下穿其5日均线，卖出。

## F 随机指标 RSV,KDJ_K，KDJ_D，KDJ_J
KDJ的计算比较复杂，首先要计算周期（n日、n周等）的RSV值，即未成熟随机指标值，然后再计算K值、D值、J值等。以n日KDJ数值的计算为例，其计算公式为
n日RSV=（Cn－Ln）/（Hn－Ln）×100
公式中，Cn为第n日收盘价；Ln为n日内的最低价；Hn为n日内的最高价。
其次，计算K值与D值：
当日K值=2/3×前一日K值+1/3×当日RSV
当日D值=2/3×前一日D值+1/3×当日K值
若无前一日K 值与D值，则可分别用50来代替。
J值=3*当日K值-2*当日D值
以9日为周期的KD线为例，即未成熟随机值，计算公式为
9日RSV=（C－L9）÷（H9－L9）×100
公式中，C为第9日的收盘价；L9为9日内的最低价；H9为9日内的最高价。
K值=2/3×第8日K值+1/3×第9日RSV
D值=2/3×第8日D值+1/3×第9日K值
J值=3*第9日K值-2*第9日D值
** 注意 **
优矿提供的数据可能有问题。因为K值，D值理论上处在0-100之间。
** 简单交易示例 **
K上穿D，则买入，
下穿则清空

## G 变动率指标 ROC
ROC=(今天的收盘价-N日前的收盘价)/N日前的收盘价*100
** 注意 **
ROC指标过于敏感，短期波动频率太快
** 改进方案：**
成交均价=个股成交额/个股成交量
ROC=(今天的成交均价-N日前的成交均价)/N日前的成交均价*100

** 简单交易示例 **
ROCEMA=[2*ROC+(X-1)*上一周期ROC]/(X+1)
ROC 上穿ROCEMA则买进。
ROCEMA 下穿ROCEMA则清空。

## H. 相对离散指数 RVI，UpRVI，DownRVI


RVI = 100 $\times$ UpRVI/(UpRVI + DownRVI) 类似 RSI

如果 close price 大于昨日，今日的上涨幅度为u: u = STD (n1).即前n1日的标准差
如果 close price 小于昨日，今日的下降幅度为d: d = STD (n1).即前n1日的标准差

UpRVI = sum (u)/n2.为前n2日所有的上涨的均值
DownRVI = sum (d)/n2.为前n2日所有的下降的均值



## I. [钱德动量摆动指标](https://uqer.datayes.com/community/share/5590d8bdf9f06cb5604f1881)（Chande Momentum Osciliator)


- $CMO_n$是一个n天滚动指标，在这n天中的第i天计算每天的 收盘价 - 前收盘价 ，如果为正则赋给$up_i$（$dn_i$为0），为负则将绝对值赋给$dn_i$（$up_i$为0）
$$ChandeSU = \sum_{i=1}^{n}{up_i}$$
$$ChandeSD = \sum_{i=1}^{n}{dn_i}$$
- 其计算公式为：

$$CMO_n=\frac{ChandeSU-ChandeSD}{ChandeSU+ChandeSD}*100$$

** 策略展示 **
因为优矿提供的数据没有说明日期，具体代码我们自己实现

CMO 小于0且大于-80   或 大于90 清空
参数windowns 是计算公式中的n
