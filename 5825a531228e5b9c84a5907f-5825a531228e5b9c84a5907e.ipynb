一. 卡方分布：k个服从标准正太分布的独立随机变量的平方和服从自由度为k的卡方分布

可以通过多次随机抽取df个正太分布变量来对比

```python
import numpy as np
import scipy.stats as sps 
import matplotlib.pyplot as plt 
df=5
Xnm=sps.norm.rvs(size=(df,100000))
Xm=np.sum(Xnm*Xnm,axis=0)
plt.hist(Xm,bins=100,normed=True,color='b')

x=np.linspace(start=0, stop=20,num=100)
plt.plot(x,sps.chi2.pdf(x,df),'r')
plt.legend(['sum_norm2','chi2'])
plt.xlabel('x')
plt.ylabel('pdf')
plt.show()
```

对比自由度不同的卡方分布

```python
import numpy as np
import scipy.stats as sps 
import matplotlib.pyplot as plt 
x=np.linspace(start=0, stop=20,num=100)
plt.plot(x,sps.chi2.pdf(x,df=1))
plt.plot(x,sps.chi2.pdf(x,df=2))
plt.plot(x,sps.chi2.pdf(x,df=3))
plt.plot(x,sps.chi2.pdf(x,df=4))
plt.plot(x,sps.chi2.pdf(x,df=5))
plt.plot(x,sps.chi2.pdf(x,df=10))
plt.legend(['df=1','df=2','df=3','df=4','df=5','df=10'])
plt.show()
```

两个服从卡方分布的相互独立的随机变量的和服从自由度为两者之和的卡方分布

```python
import numpy as np
import scipy.stats as sps 
import matplotlib.pyplot as plt 
df1=2
Xn1=sps.chi2.rvs(df1,size=100000)
df2=3
Xn2=sps.chi2.rvs(df2,size=100000)
x=np.linspace(start=0, stop=20,num=100)
plt.hist(Xn1+Xn2,bins=100,normed=True,color='b')
plt.plot(x,sps.chi2.pdf(x,df1+df2),'r')
plt.legend(['sum_chi2','chi2'])
plt.xlabel('x')
plt.ylabel('pdf')
plt.show()
```

二. t分布：A服从标准正太分布,B服从自由度为df的卡方分布，且A、B相互独立，则A/sqrt（B/df）服从t分布

```python
import numpy as np
import scipy.stats as sps 
import matplotlib.pyplot as plt 
Xn1=sps.norm.rvs(size=100000)
df=9
Xn2=sps.chi2.rvs(df,size=100000)
x=np.linspace(start=-5, stop=20,num=100)
plt.hist(Xn1/np.sqrt(Xn2/df),bins=100,normed=True,color='b')
plt.plot(x,sps.t.pdf(x,df),'r')
plt.legend(['norm/chi2','t'])
plt.xlabel('x')
plt.ylabel('pdf')
plt.show()
```

当n很大(n>30)时，t分布趋于标准正太分布

```python
import numpy as np
import scipy.stats as sps 
import matplotlib.pyplot as plt 
x=np.linspace(start=-5, stop=5,num=100)
# plt.hist(sps.norm.rvs(size=10000))
plt.plot(x,sps.t.pdf(x,df=1))
plt.plot(x,sps.t.pdf(x,df=5))
plt.plot(x,sps.t.pdf(x,df=10))
plt.plot(x,sps.t.pdf(x,df=50))
plt.plot(x,sps.norm.pdf(x))
# plt.plot(x,sps.norm.pdf(x,loc=0,scale=0.5))
# plt.plot(x,sps.norm.pdf(x,loc=1,scale=1))
plt.legend(['df=1','df=5','df=10','df=50','norm'])
plt.show()
```

三，F分布：A、B是分别服从自由度为df1、df2的卡方分布，则(A/df1)/(B/df2)服从F（df1，df2）分布

```python
import numpy as np
import scipy.stats as sps 
import matplotlib.pyplot as plt 
df1=30
Xn1=sps.chi2.rvs(df1,size=100000)
df2=50
Xn2=sps.chi2.rvs(df2,size=100000)
x=np.linspace(start=0, stop=20,num=100)
plt.hist((Xn1/df1)/(Xn2/df2),bins=30,normed=True,color='b')
plt.plot(x,sps.f.pdf(x,df1,df2),'r')
plt.legend(['chi2/chi2','f'])
plt.xlabel('x')
plt.ylabel('pdf')
plt.show()
```

若A服从t（df）分布，则A*A服从F（1，df）分布

```python
import numpy as np
import scipy.stats as sps 
import matplotlib.pyplot as plt 
df=50
Xn=sps.t.rvs(df,size=100000)
x=np.linspace(start=0, stop=20,num=100)
plt.hist(Xn*Xn,bins=80,normed=True,color='b')
plt.plot(x,sps.f.pdf(x,1,df),'r')
plt.legend(['t(n)*t(n)','f(1,n)'])
plt.xlabel('x')
plt.ylabel('pdf')
plt.show()
```
