```python
import pandas as pd
import numpy as np
import DataAPI
import matplotlib
from CAL.PyCAL import *
cal = Calendar('China.SSE')
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import matplotlib.patches as patches
import datetime 
from dateutil.parser import parse
import talib as ta
from scipy.signal import argrelextrema
from numpy import *
from quartz_futures.api import *


universe = set_universe('A')
begin = '20170320'
end = '20170520'
```

```python
def seek_max_up_stk(begin_date,end_date,universe,up,down,num=0):
    #参数：begin_date='20170320'  end='20170420' universe=set_universe('A')
    #up=0.9 down=0.8 up,down分别为时间段内涨幅最厉害的90%至80%
    #num为取涨幅最厉害的num只股票，updown与num两者选一
    dit ={}
    for stk in universe:
        tmp_data = DataAPI.MktEqudGet(tradeDate=u"",secID=stk,ticker=u"",beginDate=begin_date,endDate=end_date,isOpen="",field=u"tradeDate,chgPct",pandas="1").set_index('tradeDate').sort_values('chgPct',ascending=False)
        dit[stk] = sum(tmp_data.chgPct)
    all_df = pd.DataFrame(index=dit.keys(),data=dit.values(),columns=['zhangdie'])
    if up and down and num ==0:
        max_up_stk = all_df[all_df.zhangdie > all_df.zhangdie.quantile(down)][all_df.zhangdie < all_df.zhangdie.quantile(up)].sort_values('zhangdie',ascending=False)
        max_up_stk = max_up_stk.index.values
    if num !=0 :
        max_up_stk = all_df.sort('zhangdie')
        max_up_stk = max_up_stk.index[:num].values
    return max_up_stk

#pd.sort_values(列名，正序或倒序) 第九行
```
