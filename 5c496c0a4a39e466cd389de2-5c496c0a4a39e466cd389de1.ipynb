```python
"""
    单个股票当前的内在价值模型(资产评估法)
"""
#引入相关模块
import numpy as np
import pandas as pd
import math
from datetime import datetime, timedelta
import matplotlib.pyplot as plt
import talib as ta 
from CAL.PyCAL import * 
import matplotlib.colors as colors
import matplotlib.cm as cmx
import random as rd
#获取资产负债表财务信息
balanceSheetData = DataAPI.FdmtBSGet(ticker=u"600611",secID=u"",reportType=u"",endDate=u"2018-12-31",beginDate=u"2018-09-01",pandas="1")
#获取总股本数据
totalCapitalStock = DataAPI.EquShareGet(secID=u"",ticker=u"600611",beginDate=u"2018-06-30",endDate=u"2018-12-30",field=u"",pandas="1")
#总资产
totalAssetsData = balanceSheetData.TAssets
#总负债
totalLiablitiesData = balanceSheetData.TLiab
#获取总股本
capitalStockData = totalCapitalStock.totalShares
#计算600611(大众交通)的内在价值——资产评估法
intrinsicValue = (totalAssetsData-totalLiablitiesData)/capitalStockData
#第二步：获取当前股票的昨日收盘价信息
stockPrice = DataAPI.MktEqudGet(secID=u"",ticker=u"600611",tradeDate=u"2019-01-23",beginDate=u"",endDate=u"",isOpen="",field=u"",pandas="1")
closePrice = stockPrice.closePrice

overflowRatio = (closePrice-intrinsicValue)/intrinsicValue*100

#将数据转换为dataFrame
data = pd.DataFrame({u'内在价值':intrinsicValue,u'溢出比率':overflowRatio})

print(data)
```
