# 一.测度变换
- ### 人们常说p-quant,q-quant,为什么如此说呢？其实是来自于测度变换
- ### 我们举个简单的例子来说明一下测度变换是如何进行的：
- #### 我们知道概率空间由三部分内容构成$(\Omega,F,P)$。
- #### 假设样本空间$\Omega=\{\omega_1,\omega_2\}$,并且概率测度P满足：$p(\omega_1)=p,p(\omega_2)=1-p$
- #### 下面定义一个等价的概念，如果两个概率测度有相同的零测集(P(A)=0等价于Q(A)=0),那么它们就是等价的，记作：$Q \sim P$
- #### 比如在这里，可以令Q是一个新的等价于P的测度，满足： $Q(\omega_1)=q,Q(\omega_2)=1-q,0 < q < 1$
- #### 这时候我们还可以定义一个算子来描述这个测度变换，$\Lambda = \frac{Q}{P}$,即，$\Lambda(\omega_1) = \frac{Q(\omega_1)}{P(\omega_1)} = \frac{q}{p}$,$\Lambda(\omega_2) = \frac{Q(\omega_2)}{P(\omega_2)} = \frac{q}{p}$
- #### 通过推导，这个算子还满足一些性质：
> - #### $E_P(\Lambda)=1$
> - #### 随机变量X在Q下的期望为：$E_Q(X)=E_P(\Lambda X)$
> - #### 对于计算概率而言，$Q(x\in A)=E_P(\Lambda I(x\in A))$
 
- #### 从这个例子可以看出，所谓测度变换，变换的只是概率测度而已 

## **正态随机变量的测度变换**
- ### 刚刚是一个简单的离散的情形，下面看对于连续随机变量的测度变换是怎样的:
- ### 假设X服从正态分布$N(\mu,1)$，那么密度可以表示为：
- ### $$f_{\mu}(x)=\frac{1}{\sqrt{2\pi}}e^{-\frac{1}{2}(x-\mu)^2}=f_0(x)e^{\mu x-\mu^2/2}$$
- ### 这时候我们可以令$\Lambda(x)=e^{\mu x-\mu^2/2}$，即$f_{\mu}(x)=f_0(x)\Lambda(x)$
- ### 通过推导，可以证明这个算子是满足以上要求和性质的，也就是说
- ### 我们可以把$N(\mu,1)$看作是从$N(0,1)$测度变换而来的，把以上整理一下描述如下：
> ### 如果随机变量X在测度P下有正态分布$N(0,1)$,并$\Lambda(X)=e^{\mu X-\mu^2/2}$,那么可以定义测度Q：
> ### $\frac{dQ}{dP}(X)=\Lambda(X)$,则Q是一个和P等价的测度，并且随机变量在Q下的分布为$N(\mu,1)$

## **一般空间的测度变换**
- ### 假设$P$和$Q$是定义在同一空间上的概率测度，我们再定义一个比等价弱一点的条件：
- ### 如果P(A)=0就能推出Q(A)=0，那么就说$Q\ll P$;如果$Q\ll P$且$P\ll Q$,那么可以说$P\sim Q$
- ### 下面介绍两个定理，保证了测度变换的可行性：
> ### **Radon-Nikodym**
> #### 如果$Q\ll P$，那么必存在一个随机变量$\Lambda$,使得，$\Lambda \geq 0,E_P\Lambda=1$,并且$Q(A)=E_p(\Lambda I(A))=\int_{A} \Lambda dP$
> ### **general Bayes formula.**
> #### 设G是P，Q所定义的概率空间上的$\sigma$代数域F的子$\sigma$代数域，如果有$Q\ll P，dQ=\Lambda dP$，X在Q下是可积的，则$\Lambda X$是P下可积的，并且在Q下以概率1成立$E_Q(X|G)=\frac{E_P(X\Lambda|G)}{E_P(\Lambda|G)}$

# 二.鞅
- ### 关于布朗运动的定义还是需要在这里重复一下：
> - ### $B(t)-B(s)$与$B(u)，0\leq u\leq s$独立
> - ### $B(t)-B(s) \sim N(0,t-s)，s<t$　
> - ### $B(t)$轨道连续

- ### 关于随机过程，还有一个定义，就是关于过滤的$(F_t)$,数学上它是一个$\sigma$代数域，直观的理解的话，$F_t$就是t时刻前所有的市场信息。
- ### 下面介绍关于鞅的定义.对于随机过程$M(t),0\leq t\leq T$(连续或离散)，如果它是鞅，需要满足以下条件：
> - ### $E|M(t)|< \infty$
> - ### $E(M(t)|F_s)=M(s),\forall s< t$

- ### 直观理解就是希望这个过程的期望是平稳的常值，在s时刻所知道的所有信息下整个过程的期望就等于s时刻的值，所以这是一个公平的过程，没有偏向上涨或下跌的趋势
- ### 可以证明，布朗运动就是一个鞅

# **三.等价鞅测度（ＥＭＭ）**
- ### 通过前面的介绍，可以猜到，所谓的等价鞅测度，其实就是找一个测度使得某个过程在该测度下是一个鞅
- ### 通常情况下，我们找的那个测度Q是使得股价的折现过程$Z(t)=S(t)/\beta(t)$是一个Ｑ下的鞅，这样的测度Q我们称之为`等价鞅测度`或者`风险中性测度`
> ## 第一基本定理
> #### 市场无套利机会等价于存在风险中性测度（即使得$Z(t)=S(t)/\beta(t)$在该测度下是一个鞅）

- ### 通常$\beta(t)=e^{rt}$
- ### 比如在经典的Black-Scholes 模型中，我们假设股价是如下的一个扩散过程：
$$dS(t)=\mu S(t)dt+\sigma S(t)dB(t)$$
- ### 通过求解这个随机微分方程，我们可以得到$Z(t)=S(t)e^{-rt}=S(0)e^{\mu-r-\frac{1}{2}\sigma^2t+\sigma B(t)}$
- ### 这个过程可以证明只有在$\mu=r$时才是鞅.那么我们就需要找一个测度Q使得：
- ### $$\mu dt+\sigma dB(t)=r dt+\sigma dW(t)$$
- ### 其中，$W(t)$是一个Ｑ下的布朗运动，这个Ｑ就是ＥＭＭ，这时候可以写为：
- ### $$S(t)=S(0)e^{r-\frac{1}{2}\sigma^2t+\sigma W(t)},Z(t)=S(0)e^{-\frac{1}{2}\sigma^2t+\sigma W(t)}$$
- ### 从数学上，可以证明，这个Ｑ是唯一存在的，$\Lambda$算子的形式如下：
- ### $$\Lambda=\frac{dQ}{dP}=e^{\int_{0}^{T}H(t)dB(t)-\frac{1}{2}\int_{0}^{T}H^2(t)dt}$$
- ### 其中,$H(t)=\frac{\mu(S(t))-r(t)S(t)}{\sigma(S(t))}$
