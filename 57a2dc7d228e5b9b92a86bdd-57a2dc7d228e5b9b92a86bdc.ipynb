# **融资融券介绍**
- ## 融资融券
融资融券交易，是指投资者向具有融资融券业务资格的证券公司提供担保物，借入资金买入证券（融资交易）或借入证券并卖出（融券交易）的行为。包括券商对投资者的融资、融券和金融机构对券商的融资、融券。
- ## 影响
融资融券交易,不仅为我国股市引入杠杆交易，更重要的是，结束了我国个股不能做空的历史，使得对某股票持负面看法的投资者，能够在一定程度上表达其观点。与普通交易者不同的是，融资融券交易者为杠杆交易者，他们不仅需要承担一定的融资利率或融券费率，而且还要承担追加保证金的潜在压力。
经典理论认为，融券交易相当于增加了股票的供给渠道，使得供给曲线向右平移，从而降低股票价格；融资交易使得需求曲线向右平移，从而推高股票价格。因此`股票究竟下跌还是上涨，应当取决于融资融券交易的相对力度`，如果融资交易逐步走强，则股票可能以上涨为主，如果融券交易逐步走强，则股票可能以下跌为主。

# **纳入融资融券标的事件影响**
- 我们不妨先研究一下股票被纳入融资融券标的这一事件所带来的投资机会。
- 主要的方法是观察股票被纳入融资融券标的以后几个交易日内的收益状况。
- 均值反应的是整体的趋势，我们先从均值的角度来观察究竟在事件发生后持有几个交易日的收益最高：

```python
#加载包
import numpy as np
import pandas as pd
from matplotlib import pylab
import matplotlib.pyplot as plt
from matplotlib.pyplot import *
matplotlib.style.use('ggplot')
import seaborn as sns

from CAL.PyCAL import *
font.set_size(22)
```

```python
#处理日期，取交易日日期存在date里
data = DataAPI.TradeCalGet(exchangeCD=u"XSHG",beginDate='20070101',endDate='20160731',field=['calendarDate','isOpen'])
data = data[data['isOpen'] == 1]
date = map(lambda x: x[0:4]+x[5:7]+x[8:10], data['calendarDate'].values.tolist())
```

```python
#取A股中所有被纳入融资融券标的的股票
stklist=DataAPI.FstDetailGet(secID=set_universe('A'),beginDate=u"20100101").secID.unique().tolist()
```

- 接下来提取融资融券的数据，即被每只股票被纳入的时间（有些股票被纳入后被剔除，然后又被纳入，这样也算一次）

```python
#这个API需要去商城开通试用
liangrongdata = DataAPI.JY.FstTargetSecGet(secID=stklist)   #取stklist中股票的两融数据
liangrongdata = liangrongdata[['secID','intoDate','targetTypeCD']]
liangrongdata = liangrongdata[liangrongdata['targetTypeCD'] == 1]
liangrongdata = liangrongdata[['secID','intoDate']]
liangrongdata['intoDate'] = map(lambda x: x[0:4]+x[5:7]+x[8:10], liangrongdata['intoDate'].values.tolist())  #处理一下日期成我们熟悉的形式，如：20150101
```

- 看一下取出来的数据

```python
liangrongdata.head(10)  #前十个数据
```

- 以纳入标的日的第二天买入，持有十个交易日为例子，计算一下每只股票的收益和收益的均值是怎样的

```python
buydata = pd.DataFrame()   #存收益的数据
buydata['secID'] = liangrongdata['secID']     #取股票ID
buydata['buydate'] = map(lambda x: date[date.index(x)+1], liangrongdata['intoDate'].values.tolist())  #纳入标的的第二天的交易日
buydata['selldate'] = map(lambda x: date[date.index(x)+11], liangrongdata['intoDate'].values.tolist())  #持有十天后的交易日
buydata['buyprice'] = np.zeros(len(buydata)) 
buydata['sellprice'] = np.zeros(len(buydata))
for i in range(len(buydata)):
    buydata.iat[i,3] = DataAPI.MktEqudAdjGet(tradeDate=buydata.iat[i,1],secID=buydata.iat[i,0],field='closePrice').iat[0,0]  #获取买入时的前复权收盘价
    buydata.iat[i,4] = DataAPI.MktEqudAdjGet(tradeDate=buydata.iat[i,2],secID=buydata.iat[i,0],field='closePrice').iat[0,0]  #获取卖出时的前复权收盘价
buydata['return'] = buydata['sellprice']/buydata['buyprice']-1  #存收益
```

```python
buydata.tail(10)
```

- 可以看到，这样比较直观，我们不妨算一下收益的均值

```python
buydata['return'].mean()
```

- 那么，持有十个交易日是否是最好的呢？我们不妨把刚才的过程做成一个函数，然后改一下持有的长度，再来观察一下结果：

```python
def get_returnMean(N):   #传入的是持有的天数
    buydata = pd.DataFrame()
    buydata['secID'] = liangrongdata['secID']
    buydata['buydate'] = map(lambda x: date[date.index(x)+1], liangrongdata['intoDate'].values.tolist())  #纳入标的的第二天的交易日
    buydata['selldate'] = map(lambda x: date[date.index(x)+1+N], liangrongdata['intoDate'].values.tolist())  #持有N天后的交易日
    buydata['buyprice'] = np.zeros(len(buydata))
    buydata['sellprice'] = np.zeros(len(buydata))
    for i in range(len(buydata)):
        buydata.iat[i,3] = DataAPI.MktEqudAdjGet(tradeDate=buydata.iat[i,1],secID=buydata.iat[i,0],field='closePrice').iat[0,0]
        buydata.iat[i,4] = DataAPI.MktEqudAdjGet(tradeDate=buydata.iat[i,2],secID=buydata.iat[i,0],field='closePrice').iat[0,0]
    buydata['return'] = buydata['sellprice']/buydata['buyprice']-1
    return buydata['return']  #返回收益的序列
```

- 取持有天数从1到30天：

```python
holddays = pd.DataFrame()
holddays['return_mean'] = [get_returnMean(x).mean() for x in range(1,31)]  #取不同持有天数，收益的平均数
holddays.index=np.arange(1,31)
```

```python
#来自涛哥的中文画图函数
def fig_style(ax, legend, legend_loc='upper right', xlabel=''):
    """自定义图形属性及格式：
    ax: 图形对象
    legend: 标注
    legend_loc: 标注位置
    xlable: x 坐标名称
    """
    ax.legend(legend, prop=small_font, loc=legend_loc)
    ax.title.set_font_properties(font)
    if xlabel:
        ax.set_xlabel(xlabel, fontproperties=small_font)
    ax.set_axis_bgcolor('white')
    ax.grid(color='gray', alpha=0.2, axis='y')
    return ax
```

```python
ax = holddays.plot(linestyle='--',figsize=(14, 7), title=u'平均收益随持有天数变化')
fig_style(ax, [u'平均收益'], 'best', xlabel=u'持有天数')
```

```python
holddays.T
```

- 可以看到持有8天，平均收益是最高的，十天以后，平均收益就开始往下走了。
- 再来看看持有8天时收益的分布状况：

```python
hold8days = pd.DataFrame()     
hold8days['return'] = get_returnMean(8)   #获取持有8天时的数据
```

```python
ax = hold8days.plot(kind='hist',figsize=(14, 7),color='r',title=u'持有8天的收益分布')
fig_style(ax, [u'收益分布'], 'best', xlabel=u'收益')
```

- 从频数上来讲，大部分收益都在0~10%之间
- 再看看，收益在均值之上的股票的个数是多少：

```python
print '总股票的个数:'+str(len(hold8days))
print '收益在均值之上的股票的个数:'+str(len(hold8days[hold8days['return']>hold8days['return'].mean()]))
print '收益大于0的股票的个数:'+str(len(hold8days[hold8days['return']>0]))
```

- 收益大于0的股票的个数达到了60%左右
- 收益在均值之上的股票的个数还没有超过一半，看来我们应该取中位数更准确一些。
- 下面用`中位数`来重复一遍上述过程，结论就直接放下面了。

```python
holddaysd = pd.DataFrame()
holddaysd['return_median'] = [get_returnMean(x).median() for x in range(1,31)]  #取不同持有天数，收益的平均数
holddaysd.index=np.arange(1,31)
```

```python
holddaysd.max()
```

```python
holddaysd.T
```

- 好吧，结论还是持有8个交易日最好。

# **融资融券交易的影响**
- 上面已经说过，股票究竟下跌还是上涨，应当取决于融资融券交易的相对力度
- 我们先定义一个衡量融资融券相对力度的指标-融资余额比：
$$SMTR=融资余额/(融资余额+融券余额)$$
- 从原理上来说，我们可以认为融资余额比上升时，股票被看好，上涨的可能大；反之同理。

- 我们来看一下以这个指标来选股，是否有好的表现
