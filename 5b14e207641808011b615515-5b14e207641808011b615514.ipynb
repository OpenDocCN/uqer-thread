# pandas学习：日期处理的一些方法

```python
import pandas as pd
import time
import datetime
```

1 时间戳的构建

1.1 Timestamp()的构造方法

```python
p1=pd.Timestamp(2017,6,19)
p2=pd.Timestamp(datetime.datetime(2017,6,19,hour=9,minute=13,second=45))
p3=pd.Timestamp("2017-6-19 9:13:45")

print type(p1)
print p1
print type(p2)
print p2
print type(p3)
print p3
```

1.2  to_datetime()方法

```python
p4=pd.to_datetime("2017-6-19 9:13:45")
p5=pd.to_datetime(datetime.datetime(2017,6,19,hour=9,minute=13,second=45))

print('type of p4:',type(p4))
print(p4)
print('type of p5:',type(p5))
print(p5)
```

使用timedelta寻找昨天和明天的日期

datetime模块的对象有如下：
timedelta
date
datetime
time
tzinfo

```python
p6=p5+datetime.timedelta(days=-1)
p7=p5+datetime.timedelta(days=1)
p8=p5+datetime.timedelta(days=2)

print(type(p6))
print("Today is:",p5)
print("Yesterday is:",p6)
print("Tomorrow is:",p7)
print("The day after tomorrow is:",p8)
```

可以查看时间戳的年月日等数据

```python
print(p8.year)
print(p8.month)
print(p8.day)
print(p8.hour)
print(p8.minute)
print(p8.second)
```

3 timestamp、datetime、string之间的转换

3.1 Timestamp转datetime

```python
# 转化成datetime类型
p9=p8.to_datetime()
# 转化成date类型
p10=p8.date()

print('type of p8:',type(p8))
print(p8)
print('type of p9:',type(p9))
print(p9)
print('type of p10:',type(p10))
print(p10)
```

3.2 datetime转string

```python
p11=p9.strftime("%Y%m%d")
print('type of p11:',type(p11))
print(p11)
p12=p9.strftime("%Y-%m-%d")
print('type of p12:',type(p12))
print(p12)
```
