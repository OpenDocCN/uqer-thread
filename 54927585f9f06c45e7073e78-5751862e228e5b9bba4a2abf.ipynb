###散户之家荐股

这是利用“散户之家”推荐来买卖股票的策略

具体为：

>* 买入：若“散户之家”推荐，则买入该股票
    
>* 卖出：若股票持有3天，则卖出

```python
#读取文件，将文件内容处理，hold_days=3

import pandas as pd
from datetime import datetime,timedelta 
from CAL.PyCAL import *
a=pd.read_excel('SHZJ.xlsx','Sheet1')                 # 读取excel文件
c = a[a.tickers.notnull() | a.extra_tickers_from_name.notnull()]   #如果tickers和extra_tickers_from_name都missing，则删去这一行
b = c[['extra_tickers_from_name', 'tickers','time']]
#用stock_list储存某一天应该买入的股票
buy_list={}
#建立一个列表universe_new，储存投资域
universe_new=[]
#建立一个字典stock_dic，将ticker与secID一一对应            
stock_dic={} 
#推荐的代码有误，有的不存在对应股票，或者是基金，水星不支持
error_stk=[]
for i in range(len(b)):
#for i in range(100):

    #将b中‘date’一项的数据取出来
    date=b.iloc[i,2][0:10]
    #将b中‘tickers’一项的数据取出来
    s1=b.iloc[i,1]
    #将b中'extra_tickers_from_name'一项的数据取出来
    s2=b.iloc[i,0]

    if pd.notnull(s1):
        l1=s1.split(' ')
    else:
        l1=[]
    if pd.notnull(s2):
        l2=s2.split(' ')
    else:
        l2=[]
    l_all=l1+l2
    for j in l_all:
        try:
            stock_new=DataAPI.SecIDGet(ticker=j).loc[0,'secID']
            if stock_new not in universe_new:
                stock_dic[j]=stock_new
                universe_new.append(stock_new)
        except:
            error_stk.append(j)
            l_all.remove(j)
    
    if date not in buy_list:
        buy_list[date]=l_all
    else:
        buy_list[date] += l_all
    
```
