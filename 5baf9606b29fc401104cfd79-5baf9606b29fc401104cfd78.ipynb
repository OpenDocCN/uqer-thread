## <font color=#0099ff>一，指数轮动</font>

---
以沪深300/中证500指数，构建月度的指数轮动模型

A.**使用债券市场以及股票市场相关指标构建择时变量**
- 利率类指标，波动率类指标，估值类指标在月频调仓下对于未来1个月沪深300与中证500指数收益差具有较强的预测效果 
- 指标分别是：期限利差变动，指数波动率截面差值，指数一致预期估值截面差值

B.**使用Logistic回归进行指数相对强弱概率预测**
- 使用Logistic回归滚动预测未来1期中指数收益差为正的概率
- 概率高于55%时，做中证500；概率低于45%时，做多沪深300；概率介于45%至55%之间时，做多中证800

## <font color=#0099ff> 定义函数 </font>

```python
# 指数涨跌幅
def get_index_pct(begin, end, index, name):
    idx_pct = DataAPI.MktIdxmGet(beginDate=begin,endDate=end,indexID=index,field=['endDate', 'chgPct'],pandas="1")
    idx_pct.rename(columns={'endDate':'tradeDate', 'chgPct':name}, inplace=True)
    idx_pct['tradeDate'] = idx_pct['tradeDate'].apply(lambda x: x.replace("-", ""))
    idx_pct.set_index('tradeDate', inplace=True)
    idx_pct.sort_index(ascending=True, inplace=True)
    return idx_pct

# 指数N日涨跌幅
def get_index_pct(date_list, ticker_list, N):   
    cal = Calendar('China.SSE')
    date_list = sorted(date_list)
    ret = pd.DataFrame()
    for date in date_list:
        start = cal.advanceDate(date, '-%sB'%N).strftime('%Y%m%d')
        temp1 = DataAPI.MktIdxdGet(ticker=ticker_list,tradeDate=start,exchangeCD=u"XSHE,XSHG",field=u"tradeDate,indexID,closeIndex",pandas="1")
        temp2 = DataAPI.MktIdxdGet(ticker=ticker_list,tradeDate=date,exchangeCD=u"XSHE,XSHG",field=u"tradeDate,indexID,closeIndex",pandas="1")
        temp2['ret_' + str(N)] = temp2['closeIndex'] / temp1['closeIndex'] - 1
        ret = pd.concat([ret, temp2],axis=0)    
    ret = ret[['tradeDate', 'indexID', 'ret_' + str(N)]]
    ret = ret.pivot_table(index='tradeDate', columns='indexID', values='ret_' + str(N))
    ret['date'] = ret.index
    ret.reset_index(drop=True, inplace=True)
    ret['date'] = map(time_change, ret['date'])
    return ret

# 指数N天波动率                              
def get_index_vol(date_list, ticker, N):
    cal = Calendar('China.SSE')
    start = cal.advanceDate(date_list[0], '-%sB'%(N-1)).strftime('%Y%m%d')
    end = date_list[-1]
    temp = DataAPI.MktIdxdGet(ticker=ticker,beginDate=start,endDate=end,exchangeCD=u"XSHE,XSHG",field=u"tradeDate,CHGPct",pandas="1")
    temp['vol_' + str(ticker)] = pd.rolling_std(temp['CHGPct'], window=N)
    temp = temp[['tradeDate', 'vol_' + str(ticker)]]
    vol = temp.copy()
    vol['tradeDate'] = map(time_change, vol['tradeDate'])
    vol = vol[vol['tradeDate'].isin(date_list)]
    vol.columns = ['date', 'vol_' + str(ticker)] 
    return vol

# 指数_PB
def get_index_PB(date_list, ticker):
    cal = Calendar('China.SSE')
    start = date_list[0]
    end = date_list[-1]
    temp = DataAPI.MktIdxFactorDateRangeGet(ticker=ticker,beginDate=start,endDate=end,field=u"tradeDate,PB2",pandas="1")
    temp['tradeDate'] = map(time_change, temp['tradeDate'])
    temp = temp[temp['tradeDate'].isin(date_list)]
    temp.columns = ['date', 'PB_' + str(ticker)] 
    return temp

# 外生变量
def get_mhkq_data(date_list, file_="AAA.xls", name_='AAA'):
    temp = pd.read_excel(file_)
    temp.columns = ['periodDate', 'dataValue']
    temp.dropna(how='all', inplace=True)
    temp['periodDate'] = map(time_change, temp['periodDate'])
    temp = temp[temp['periodDate'].isin(date_list)]
    temp.sort_values(by='periodDate', inplace=True)
    temp.reset_index(drop=True, inplace=True)
    temp.columns = ['date', name_]
    return temp

# 格式转换
def time_change(x):
    y = datetime.datetime.strptime(x, '%Y-%m-%d')
    y = y.strftime('%Y%m%d')
    return y

# 月末时间list
def get_trade_list(start_date, end_date):
    cal_dates = DataAPI.TradeCalGet(exchangeCD=u"XSHG", beginDate=start_date, endDate=end_date).sort('calendarDate')
    cal_dates = cal_dates[cal_dates['isOpen'] == 1]
    month_end = cal_dates[cal_dates['isMonthEnd'] == 1]
    month_end = month_end['calendarDate'].values.tolist()
    month_end = [time_change(x) for x in month_end]
    return month_end
```

```python
# 读取data
def load_factor(file_path):
    dfs = []
    store = pd.HDFStore(file_path)
    for k in store.keys():
        dfs.append(store.get(k))
    df = pd.concat(dfs)
    df['tradeDate'] = df['tradeDate'].astype(str)
    store.close()
    return df

# 样本外的ACC与AUC
def get_test_auc_acc(file_path): 
    df = load_factor(file_path)
    df = df[df['label'] != 0]
    df.loc[:, 'predict'] = df.loc[:, 'factor'].apply(lambda x : 1 if x > 0.5 else -1)
    df_correct = df[df['predict'] == df['label']]
    mean_acc = len(df_correct) * 1.0 / len(df)
    mean_auc =  roc_auc_score(np.array(df['label']), np.array(df['factor']))
    return round(mean_acc, 2), round(mean_auc, 2)

# 样本外的ACC与AUC
def get_mean_acc_auc(model_names, file_prefix): 
    res = {}
    file_dir = '%s/'%(file_prefix)
    for model_name in model_names:
        period_perf = {}
        mean_acc, mean_auc = get_test_auc_acc('%s/%s.hdf'%(file_dir,model_name))
        period_perf['mean_acc'] = mean_acc
        period_perf['mean_auc'] = mean_auc
        res['%s'%(model_name)] = period_perf
    return pd.DataFrame(res)
```

```python
# 得到过去N个月的日期
def get_lastN_month(x, N, trade_frame):
    dframe = trade_frame[trade_frame.calendarDate < x].tail(N)
    if len(dframe) == 0:
        return np.nan
    else:
        return dframe.calendarDate.values[0]
    
# 得到N个月的日期区间    
def get_monthdate_relation(time_steps):
    trade_frame = DataAPI.TradeCalGet(exchangeCD=u"XSHG", beginDate="20070101", endDate="20171231", field=u"",
                                      pandas="1")
    trade_frame = trade_frame[trade_frame.isMonthEnd == 1]
    trade_frame['calendarDate'] = trade_frame['calendarDate'].apply(lambda x: x.replace("-", ""))
    trade_relation = trade_frame[['calendarDate', 'isOpen']]
    trade_relation['train_start'] = trade_frame['calendarDate'].apply(lambda x: get_lastN_month(x, time_steps, trade_frame))
    trade_relation['train_end'] = trade_frame['calendarDate'].apply(lambda x: get_lastN_month(x, 1, trade_frame))
    trade_relation.rename(columns={"calendarDate": 'tradeDate'}, inplace=True)
    trade_relation.dropna(how='any', inplace=True)
    return trade_relation

def get_train_date(curr_date, date_split_df):
    date_sample = date_split_df.query('tradeDate==@curr_date')
    if not date_sample.empty:
        return date_sample.iloc[0].to_dict()

# 训练集、测试集
def get_train_test(curr_date, data_df, date_split_df, is_filter=True):
    date_map = get_train_date(curr_date,date_split_df)
    train_start = date_map['train_start']
    train_end = date_map['train_end']
    train_df = data_df.query('(tradeDate>=@train_start)&(tradeDate<=@train_end)')
    if is_filter:
        train_df = train_df[train_df['label']!=0]
    test_df = data_df.query('tradeDate==@curr_date')
    return train_df, test_df

# 转换输入格式
def format_feature_label(origin_df, factors):
    feature = np.array(origin_df[factors],dtype=np.float32)
    label = np.array(origin_df['label'],dtype=np.int32)
    return feature, label
```

```python
# 保存文件
def dump_factor(df, predict_score, key_name,filename): 
    df['factor'] = predict_score
    df = df.loc[:, ['tradeDate', 'label', 'factor']]
    df.to_hdf(filename, key_name, mode='a')

# 训练模型
def get_fitted_model(train_df, factors):
    train_feature, train_label = format_feature_label(train_df,factors)
    train_start = time.time()
    model = LogisticRegression()
    model.fit(train_feature, train_label)
    return model

# 输出预测值
def get_single_period_factor(factors, train_df, test_df):
    model = get_fitted_model(train_df, factors)
    test_feature, test_label = format_feature_label(test_df,factors)
    pred_test = np.round(model.predict_proba(test_feature)[:,1],2)
    return pred_test

# 输出rolling预测值
def get_single_model_factor(factors, date_list, merged, date_split, dump_test_dir, N):
    dump_test_file = os.path.join(dump_test_dir, "LR%s.hdf"%N)
    for i in range(0, len(date_list)):
        train_df, test_df = get_train_test(date_list[i], merged, date_split)
        pred_test = get_single_period_factor(factors, train_df, test_df)
        dump_factor(test_df,pred_test,date_list[i],dump_test_file)

# 创建文件
def make_file_dir(file_dir):
    if not os.path.exists(file_dir):
        os.mkdir(file_dir)
```

```python
# plot收益曲线
def plot_ret(cpt_df):
    cpt_df_ = cpt_df.copy()
    cpt_df = cpt_df.dropna()
    sns.set_style("white")
    cpt_df['benchmark'] = (1 + cpt_df['benchmark']).cumprod() - 1
    cpt_df['strategy'] = (1 + cpt_df['strategy']).cumprod() - 1
    fig = plt.figure(figsize=(10, 4))
    fig.set_tight_layout(True)
    ax1 = fig.add_subplot(111)
    ax1.grid(axis='y', ls='--')
    ax1.plot(cpt_df.index, cpt_df['benchmark'], lw=2, color='grey', label='benchmark')
    ax1.plot(cpt_df.index, cpt_df['strategy'], lw=2, color='#00BFFF', label='strategy')
    ax1.legend(fontsize=16)

# plot对冲曲线
def plot_hedging(cpt_df):    
    cpt_df = cpt_df.dropna()
    sns.set_style("white")
    df_cum_rets = (cpt_df['strategy'] - cpt_df['benchmark'] + 1).cumprod()
    cpt_df['excess_cum'] = df_cum_rets - 1
    running_max = np.maximum.accumulate(df_cum_rets)
    underwater = -((running_max - df_cum_rets) / running_max)
    underwater.index = cpt_df.index
    fig = plt.figure(figsize=(10, 4))
    fig.set_tight_layout(True)
    ax1 = fig.add_subplot(111)
    ax2 = ax1.twinx()
    ax1.grid(axis='y', ls='--')
    ax1.fill_between(underwater.index, 0, np.array(underwater), color='grey', alpha=0.6, label='underwater')
    ax2.plot(cpt_df.index, cpt_df['excess_cum'], lw=2, color='tomato', label='hedging')
    ax2.set_title('hedging & underwater', fontsize=16)

# 回撤
def cal_maxdrawdown(data): 
    if isinstance(data, list):
        data = np.array(data)
    if isinstance(data, pd.Series):
        data = data.values   
    def get_mdd(values): 
        dd = [values[i:].min() / values[i] - 1 for i in range(len(values))]
        return abs(min(dd))
    if not isinstance(data, pd.DataFrame):
        return get_mdd(data)
    else:
        return data.apply(get_mdd)
    
# 回测指标   
def cal_indicators(df_daily_return, period, freq='daily'):
    if freq == 'daily':
        freq_ = 250
    if freq == 'monthly':
        freq_ = 12
    df_cum_value = (df_daily_return + 1).cumprod()
    res = pd.DataFrame(index=['年化收益率','年化标准差','贝塔','夏普比率','信息比率','最大回撤'], columns=
                       df_daily_return.columns, data=0.0)
    res.loc['年化收益率'] = (df_daily_return.mean() * freq_/period).apply(lambda x: '%.2f%%' % (x*100))
    res.loc['年化标准差'] = (df_daily_return.std() * np.sqrt(freq_/period)).apply(lambda x: '%.2f%%' % (x*100))
    res.loc['贝塔', 'benchmark'] = 1
    res.loc['贝塔', 'strategy'] = np.round((np.cov(df_daily_return['strategy'],df_daily_return['benchmark'])[0,1] 
                                                                      / np.cov(df_daily_return['benchmark'])), 2)
    res.loc['夏普比率'] = (df_daily_return.mean() / df_daily_return.std() * np.sqrt(freq_/period)).apply(lambda x: np.round(x, 2))
    res.loc['信息比率', 'strategy'] = np.round(((df_daily_return['strategy'].mean()-df_daily_return['benchmark'].mean()) 
                             / (df_daily_return['strategy']-df_daily_return['benchmark']).std() * np.sqrt(freq_/period)), 2)
    res.loc['最大回撤'] = cal_maxdrawdown(df_cum_value).apply(lambda x: '%.2f%%' % (x*100))
    return res

# plot数据图
def plot_(exogenous, x, y, N, title):
    data = exogenous.copy()
    data.set_index('tradeDate', inplace=True)
    data.index = pd.to_datetime(data.index)
    data[x] = pd.rolling_mean(data[x], window=N)
    data[y] = pd.rolling_mean(data[y], window=N)
    fig, ax1 = plt.subplots()
    sns.set_style("darkgrid")
    with plt.xkcd():
        plt.plot(data[x], color='#FF69B4', label=x)
        plt.grid(True)
        plt.legend(loc=2)
        ax2 = ax1.twinx()
        plt.plot(data[y], color='#1E90FF', label=y)
        plt.grid(False)
        plt.legend(loc=1)
        plt.title(title, fontproperties=font, fontsize=16)
```
