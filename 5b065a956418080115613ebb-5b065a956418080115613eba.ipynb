# 股息率

```python
############# 获取多个股票股息率的函数 ##############
def Div_P_factor(assessDate_i, stocks): #stocks股票列表的股息率=各自在评估日之前12个月的分红/各自评估日的总市值
    assessDate_i=pd.to_datetime(assessDate_i)
    yr = assessDate_i.year
    mn = assessDate_i.month
    dy = assessDate_i.day
    if mn >= 6: 
        beginPublish_Date_i = u""
        begin_Date_i = date(yr-2,12,31)
        end_Date_i = date(yr-1,12,31)
    else: 
        beginPublish_Date_i = u""
        begin_Date_i = date(yr-2,12,31)
        end_Date_i = date(yr-1,12,31)
    Div = DataAPI.EquDivGet(secID=stocks,ticker=u"",eventProcessCD=u"",exDivDate="",beginDate=begin_Date_i,endDate=end_Date_i,beginPublishDate=beginPublish_Date_i,endPublishDate=assessDate_i,beginRecordDate=u"",endRecordDate=u"",field=u"secID,perCashDiv,publishDate,recordDate,endDate",pandas="1") 
    Div['endDate'] = pd.to_datetime(Div['endDate']) 
    Div['publishDate'] = pd.to_datetime(Div['publishDate']) 
    Div = Div[Div['publishDate']>=add_Months(assessDate_i,-12,0)] #add_Months是我编写的一个计算往前推12个月的日期的函数
    stocks_DataFrame = DataFrame(stocks,columns=['secID'])
    Div = pd.merge(Div,stocks_DataFrame,how='outer',on=['secID']) 
    Div['endDate'].fillna(date(yr-1,12,31),inplace=True)
    Div.fillna(0,inplace=True)
    endDate_list = Div.endDate.drop_duplicates().tolist() 
    endDate_min = min(endDate_list)
    endDate_max = max(endDate_list)
    total_Shares = DataAPI.EquShareGet(secID=stocks,ticker=u"",partyID=u"",beginDate=endDate_min,endDate=endDate_max,field=u"secID,totalShares,changeDate",pandas="1")
    total_Shares['changeDate'] = pd.to_datetime(total_Shares['changeDate'])
    Div = pd.merge(Div,total_Shares,how='left',left_on=['secID','endDate'],right_on=['secID','changeDate'])
    Div.drop(['changeDate'],axis=1,inplace=True)
    Div['cashDiv']=Div['perCashDiv']*Div['totalShares']/10000/10000 
    Div = Div.pivot_table(values='cashDiv',index='secID',aggfunc=np.sum)
    Div = DataFrame(Div)
    mkt_Value = DataAPI.MktEqudGet(secID=stocks,ticker=u"",tradeDate=u"",beginDate=assessDate_i,endDate=assessDate_i,isOpen="",field=u"secID,tradeDate,marketValue",pandas="1")
    mkt_Value = mkt_Value.set_index(['secID','tradeDate']).unstack()
    mkt_Value = mkt_Value/10000/10000 
    mkt_Value.columns = mkt_Value.columns.droplevel() 
    mkt_Value.rename(columns={str(date(yr,mn,dy)):'mktValue'}, inplace = True)
    Div = Div.join(mkt_Value,how='left')
    Div['Div_P%'] = Div['cashDiv']/Div['mktValue']*100
    Div.drop(['cashDiv','mktValue'],axis = 1,inplace=True)
    return Div
```
