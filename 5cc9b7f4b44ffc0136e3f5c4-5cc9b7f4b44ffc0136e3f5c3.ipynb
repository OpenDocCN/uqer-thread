```python
floatsh_df = DataAPI.EquFloatShTenGet(secID=u"",ticker=All_A_tks_list,field=u"",pandas="1")
floatsh_df = floatsh_df.sort_values(['ticker','shNum','endDate'],ascending=[True,True,False]).drop_duplicates(['ticker','shNum'], keep='first')
print (len(floatsh_df))
print (len(floatsh_df.ticker.unique()))
print (floatsh_df.head(20).to_html())
```

2、获取最新公布的前十股东相应endDate最新的市值信息，DataAPI.MktEqudGet
--


```python
# 获取最近的交易日
def lastopen_date(tradedate):
    cal_date = datetime.strptime(tradedate,"%Y-%m-%d").date()

    if cal.isBizDay(cal_date) :    # 如果是交易日，则      
        date = cal_date
    else:                          # 如果当天不是交易日，则获得前一个交易日
        cal_wd = cal.advanceDate(cal_date, '-1B', BizDayConvention.Following)    # Date格式
        date = cal_wd   # date格式        
        # date = cal_wd.toDateTime()    # datetime格式        

    date_str = date.strftime("%Y-%m-%d")
    return date_str

print lastopen_date(tradedate)
```

```python
mktvalue_df = DataAPI.MktEqudGet(secID=u"",ticker=All_A_tks_list,tradeDate=lastopen_date(enddate),isOpen="",field=u"",pandas="1")

print (len(mktvalue_df))
print (mktvalue_df.head().to_html())
```

3、merge前十股东信息与对应日期的市值信息，研究沪深300谁买的最多
--

```python
df_merge = floatsh_df.merge(mktvalue_df[['ticker','marketValue']],on='ticker',how='left')

secid300 = '000300.ZICN'
index_detail_300 = DataAPI.IdxConsGet(secID=secid300,ticker=u"",isNew=u"",pandas="1")
set300 = set(index_detail_300['consTickerSymbol']) 
prd_ticker_list = list(set300)

df_merge = df_merge[df_merge.ticker.isin(prd_ticker_list)]
df_merge['holdvalue'] = df_merge['holdPct']*df_merge['marketValue']/100
columns = [u'ticker', u'secShortName',  u'endDate',  u'shNum', u'holdPct',  u'marketValue', u'holdvalue', u'shName']
df_merge = df_merge[columns]
print (len(df_merge))
print (len(df_merge.shName.unique()))
print (df_merge[:10].to_html())
```

```python
calmost_value = df_merge['holdvalue'].groupby(df_merge['shName']).sum()
shName_NO = df_merge.groupby('shName').size()

df_concat = pd.concat([calmost_value,shName_NO],axis=1)
df_concat.columns = ['holdvalue','shName_NO']
df_concat['endDate'] = enddate
print (df_concat.holdvalue.sum(axis=0))
print (df_concat.sort_values('holdvalue',ascending=False).head(30).to_html())
```

```python
# 时间更慢，但相比于直接全部获取需要的内存应该会小些
def GetSecID(tk_list,**kargs):    
    num = 100
    cnt_num = len(tk_list)/num
    if cnt_num > 0:
        df = pd.DataFrame({})
        for i in range(cnt_num):
            sub_df = DataAPI.SecIDGet(ticker=tk_list[i*num:(i+1)*num],**kargs)
            df = pd.concat([df,sub_df])
        if (i+1)*num != len(tk_list):
            sub_df = DataAPI.SecIDGet(ticker=tk_list[(i+1)*num:],**kargs)
            df = pd.concat([df,sub_df])
    else:
        df = DataAPI.SecIDGet(ticker=tk_list,**kargs)
    return df
```

```python
#获取最近的一个交易日，返回的是datetime格式
def CountTime():        
    today = datetime.today()
    today_str = today.strftime("%Y%m%d")
    cal_date = Date.fromDateTime(today)       
    time1=" 15:05:00"
    ben_time = datetime.strptime(today_str+time1,"%Y%m%d %H:%M:%S")
    if cal.isBizDay(cal_date) & (today>ben_time):    #如果是交易日，则判断当天是不是在15点前        
        date = today
    else:    #如果当天不是交易日，则获得前一个交易日
        cal_wd = cal.advanceDate(cal_date, '-1B', BizDayConvention.Following)    #Date格式
        date = cal_wd.toDateTime()    #datetime格式        
    return date
```
