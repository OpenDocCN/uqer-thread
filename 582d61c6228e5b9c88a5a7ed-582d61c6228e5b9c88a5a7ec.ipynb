# 一.前言
- ### 相信大家对CAPM模型都不陌生，其经典之处我也不用多说了，关于CAPM的理论部分，感兴趣的可以查看链接[CAPM](http://wiki.mbalib.com/wiki/%E8%B5%84%E6%9C%AC%E8%B5%84%E4%BA%A7%E5%AE%9A%E4%BB%B7%E6%A8%A1%E5%9E%8B)
- ### 简单地说，就是下面这个公式,即单个投资组合的收益率等于无风险收益率与市场组合风险溢价的和.
$$E(R(t)) = r_f(t)+\beta (E(R_M(t))-r_f(t))$$

- ### 上式成立，当然是有许多假设的，所以问题就来了，在市场中是否满足这些假设，CAPM是否成立呢？
- ### 关于CAPM的实证的论文很多，社区里也有一些很好的实证帖子，大致上就是用回归的方法，对下面的回归模型进行实证：
$$ R(t)−r_f(t)=\alpha+\beta(R_M(t)−r_f(t))+e(t)$$

- ### 我们常说的alpha，beta中性也是从这里来的，如果beta等于0，那么投资组合的收益就不受市场的影响，当然我们也希望alpha越高越好
- ### 后来Fama在此基础上提出了三因子，五因子模型。在实证因子模型的时候，方法也越来越多，考虑的模型也更加复杂，但多半都偏于解释股票收益率
- ### 本帖并非研究实证的方法，而是尝试探究挖掘CAPM模型里的因子，形成策略。

# 二.详情
- ### 本帖探究了这里面可以挖掘的四个因子：
> - ### R2:回归的判定系数，反应CAPM模型的好坏
> - ### R：相关系数，反应股票与市场的相关性
> - ### beta：反应受市场影响的程度
> - ### alpha：除去市场影响后的超额收益

- ### 探究方法：探查样本数据内，该因子的五分位组合表现
- ### 策略逻辑：每个调仓日，每只股票对前M日的数据进行时间序列回归，得到相应的因子，根据因子值大小确定分位组合，等权买入股票

```python
#加载需要使用的包
from __future__ import division
import numpy as np
import pandas as pd
import scipy.stats as st

from CAL.PyCAL import *    # CAL.PyCAL中包含font
font.set_size(22)
small_font = font.copy()
small_font.set_size(12)

import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('white')

import datetime
```

```python
class CAPM(object):
    
    def __init__(self,begindate,enddate,factor='R2',M=20,frequency='monthly',index_name='000906.ZICN',universe_name='000906.ZICN'):
        ###----传入参数----
        self.begindate = begindate  #开始时间
        self.enddate = enddate      #结束时间
        self.factor = factor        #想要探查的因子
        self.M = M+1
        self.frequency = frequency  #数据频率，只要日度，周度，月度三个选择
        self.universe_name = universe_name  #回测的动态股票池
        self.index_name = index_name
        ###----自定义参数----

        self.datelist = self.get_datelist()  #传入开始和结束日期，得到交易日列表list
        self.results_quantile = self.quick_test()
        
    
    ###得到日期列表
    def get_datelist(self):   
        begin_date = datetime.datetime.strptime(self.begindate,"%Y-%m-%d").strftime("%Y%m%d")
        end_date = datetime.datetime.strptime(self.enddate,"%Y-%m-%d").strftime("%Y%m%d")
        data=DataAPI.TradeCalGet(exchangeCD=u"XSHG",beginDate=begin_date,endDate=end_date,field=['calendarDate','isWeekEnd','isMonthEnd','isOpen'])
        data = data[data['isOpen'] == 1]
        if self.frequency == 'daily':   #按照规定的频率处理得到的交易日序列
            #self.M = 250
            date= map(lambda x: x[0:4]+x[5:7]+x[8:10], data['calendarDate'].values.tolist())
        if self.frequency == 'weekly':
            #self.M = 52
            data = data[data['isWeekEnd'] == 1]
            date= map(lambda x: x[0:4]+x[5:7]+x[8:10], data['calendarDate'].values.tolist())
        if self.frequency == 'monthly':
            #self.M = 12
            data = data[data['isMonthEnd'] == 1]
            date= map(lambda x: x[0:4]+x[5:7]+x[8:10], data['calendarDate'].values.tolist())
        return date
    
   ###快速回测
    def quick_test(self):
        start = self.begindate                       # 回测起始时间
        end = self.enddate     # 回测结束时间
        universe = DynamicUniverse(self.universe_name)     # 股票池
        benchmark = 'HS300'                       # 策略参考标准
        capital_base = 10000000                     # 起始资金
        # 日期处理相关
        cal = Calendar('China.SSE')
        period = Period('-1B')
        # ---------------回测参数部分结束----------------

        # 把回测参数封装到 SimulationParameters 中，供 quick_backtest 使用
        sim_params = quartz.SimulationParameters(start, end, benchmark, universe, capital_base, max_history_window=max_history_window)
        # 获取回测行情数据
        data = quartz.get_backtest_data(sim_params)
        # 运行结果
        results_quantile = {}
        # 调整参数快速回测
        for q in range(5):
            
            refresh_rate = 1
            
            
            def get_value(rf,prices,account):
                stk_tmp = np.array(prices)
                index_tmp = np.array(account.mkt_list)
                stk_r = stk_tmp[1:]/stk_tmp[:-1]-1-rf
                index_r = index_tmp[1:]/index_tmp[:-1]-1-rf
                R,_ = st.pearsonr(stk_r,index_r)
                if self.factor == 'R2':
                    return R**2
                if self.factor == 'R':
                    return R
                if self.factor == 'beta':
                    up = (stk_r*index_r).sum()-stk_r.sum()*index_r.sum()
                    down = (index_r**2).sum()-len(index_r)*((index_r.mean())**2)
                    beta = up/down
                    return beta
                if self.factor == 'alpha':
                    up = (stk_r*index_r).sum()-stk_r.sum()*index_r.sum()
                    down = (index_r**2).sum()-len(index_r)*((index_r.mean())**2)
                    beta = up/down
                    return np.mean(stk_r)-beta*np.mean(index_r)
            
            def initialize(account):                   # 初始化虚拟账户状态
                account.mkt_list = []
                pass
            
            def handle_data(account):                  # 每个交易日的买入卖出指令
                
    
                today = account.current_date
                today = Date.fromDateTime(account.current_date)  # 向前移动一个工作日
                yesterday = cal.advanceDate(today, period)
                yesterday = yesterday.toDateTime().strftime('%Y%m%d')
    
                index_data = DataAPI.MktIdxdGet(tradeDate=yesterday,indexID=self.index_name,field=u"closeIndex")['closeIndex'].values.tolist()
                account.mkt_list.append(index_data[0])
                
                if yesterday in self.datelist:
                    if len(account.mkt_list) < self.M:
                        return
                    else:
                        rf=DataAPI.ChinaDataInterestRateInterbankRepoGet(indicID=u"M120000068",beginDate=yesterday,\
                                                                         endDate=yesterday,field=u"dataValue")
                        rf = rf['dataValue'].values.tolist()
                        rf = rf[0]/100/250 
                        account.mkt_list = account.mkt_list[-self.M:]
                        stk_data = account.get_attribute_history('closePrice',self.M)
                        factor = pd.DataFrame(index=account.universe,data=0,columns=['value'])
                    
                        factor['value'] = map( lambda x:get_value(rf,stk_data[x],account),factor.index.tolist())
                        quantile1ow=np.percentile(factor['value'],q*20)
                        quantilehigh=np.percentile(factor['value'],(q+1)*20)
                        factor=factor[(factor['value']>=quantile1ow) & (factor['value']<=quantilehigh)]
                    
                        # 先卖出
                        sell_list = account.security_position
                        for stk in sell_list:
                            order_to(stk, 0)

                        # 再买入
                        buylist = factor.index
                        buy_list =[]
                        total_money = account.reference_portfolio_value
                        prices = account.reference_price 
                        for stk in buylist:
                            if np.isnan(prices[stk]) or prices[stk] == 0:  # 停牌或是还没有上市等原因不能交易
                                continue
                            buy_list.append(stk)
                        for stk in buy_list:
                            order_pct_to(stk,1/len(buy_list))
                        return
                else:
                    return
            
             # 把回测逻辑封装到 TradingStrategy 中，供 quick_backtest 使用
            strategy = quartz.TradingStrategy(initialize, handle_data)
            # 回测部分
            bt, perf = quartz.quick_backtest(sim_params, strategy, data=data, refresh_rate=refresh_rate)
            # commission = commission,
            # slippage = slippage

            # 保存运行结果
            results_quantile[q] = perf
        return results_quantile
    #得到指标
    def get_perf(self):
        result = pd.DataFrame()
        result['分位组合'] = ['0~20%','20~40%','40~60%','60~80%','80~100%']
        result['年化收益'] = [self.results_quantile[i]['annualized_return'] for i in range(5)]
        result['最大回撤'] = [self.results_quantile[i]['max_drawdown'] for i in range(5)]
        result['收益波动率'] = [self.results_quantile[i]['volatility'] for i in range(5)]
        result['信息比率'] = [self.results_quantile[i]['information_ratio'] for i in range(5)]
        result['夏普比率'] = [self.results_quantile[i]['sharpe'] for i in range(5)]
        return result
    ###画图
    def get_plot(self):
        cumulative_returns = pd.DataFrame()
        for i in range(5):
            cumulative_returns = pd.concat([cumulative_returns,self.results_quantile[i]['cumulative_returns']],axis=1)
        cumulative_returns.columns=['0~20%','20~40%','40~60%','60~80%','80~100%']
        fig = plt.figure(figsize = (16,6))
        ax = fig.add_subplot(111)
        cumulative_returns.plot(ax=ax,title=self.factor+u'分位组合表现')
        ax.legend(map(lambda x: x.decode('utf-8'),cumulative_returns.columns.values),prop=small_font,loc='best')
        ax.set_xlabel(u'交易日期', fontproperties=small_font)
        plt.xticks(fontproperties = small_font, rotation = 70)
        ax.title.set_font_properties(font)
        ax.set_axis_bgcolor('white')
        ax.grid(color='gray', alpha=1.0, axis='y')
```

```python
test = CAPM('2014-08-01','2016-10-01',factor='R2',M=20,frequency='daily',index_name='000906.ZICN',universe_name='000906.ZICN')
```

```python
test.get_perf()
```
