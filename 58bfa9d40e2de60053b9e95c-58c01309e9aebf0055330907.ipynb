```python
import numpy as np
import pandas as pd
import talib
import copy
```

## LLT

```python
def LLT(closePrice,par):
    cp=copy.deepcopy(closePrice)
    tmp=copy.deepcopy(closePrice)
    a=2.0/(par+1.0)
    for i in np.arange(2,len(tmp)):
        tmp[i]=(a-(a*a)/4)*cp[i]+((a*a)/2)*cp[i-1]-(a-3*(a*a)/4)*cp[i-2]+2*(1-a)*tmp[i-1]-(1-a)*(1-a)*tmp[i-2]
    return tmp
```

## WS_MACD

```python
def WS_MACD(close,fastperiod=26,slowperiod=66,signalperiod=9):
    ewma_f = pd.ewma(close, span=fastperiod)
    ewma_l = pd.ewma(close, span=slowperiod)
    tmp = pd.DataFrame([close,ewma_f, ewma_l]).T
    tmp.columns=(['close','fast','slow'])
    tmp['DIF'] = tmp['fast'] - tmp['slow']
    tmp['DEA'] = pd.ewma(tmp['DIF'], span=signalperiod)
    tmp['MACD'] = tmp['DIF'] - tmp['DEA']
    tmp['shiftclose35'] = tmp['close'].shift(64)
    tmp['MACDSN'] = tmp['MACD']/tmp['shiftclose35']
    DIF = tmp.ix[tmp.index[-1],['DIF']].values
    MD = tmp.ix[tmp.index[-1],['MACDSN']].values
    return DIF[0], MD[0]
```

# ATR通道

```python
def ATR_channel(high, low, close, timeperiod, b=3):
    ATR = talib.ATR(high, low, close, timeperiod)[-1]
    MeanLen = np.sum(close[-timeperiod:])/timeperiod
    UpLean = MeanLen + b*ATR
    DownLean = MeanLen - b*ATR
    return UpLean, MeanLen, DownLean
```
