```python
from CAL.PyCAL import *
from datetime import datetime
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rc
rc('mathtext', default='regular')
import seaborn as sns
sns.set_style('white')
from matplotlib import dates
from pandas import concat
from scipy import interpolate
import math

def histVolumeOpt50ETF(beginDate, endDate):
    optionVarSecID = u"510050.XSHG"
    cal = Calendar('China.SSE')
    dates = cal.bizDatesList(beginDate, endDate)
    dates = map(Date.toDateTime, dates)                         #转换日期格式
    columns = ['callVol', 'putVol', 'callValue',   
               'putValue', 'callOpenInt', 'putOpenInt',
               'nearCallVol', 'nearPutVol', 'nearCallValue', 
               'nearPutValue', 'nearCallOpenInt', 'nearPutOpenInt',
               'netVol', 'netValue', 'netOpenInt',
               'volPCR', 'valuePCR', 'openIntPCR', 
               'nearVolPCR', 'nearValuePCR', 'nearOpenIntPCR']
    hist_opt = pd.DataFrame(0.0, index=dates, columns=columns)  #左列为日期 上面为columns的字符串
    hist_opt.index.name = 'date' #左上角设为'date' 

    for date in hist_opt.index:
        date_str = Date.fromDateTime(date).toISO().replace('-', '') #把2015-2-9的-去掉，遍历中的data

        try:
            opt_data = DataAPI.MktOptdGet(secID=u"", tradeDate=date_str, field=u"", pandas="1") #相当于for date in hist_opt.index: opt_data = xxx 
    
        except:
            hist_opt = hist_opt.drop(date)
            continue
    #print opt_data 
        opt_type = []
        exp_date = []
        for ticker in opt_data.secID.values: #ticker=510050C1503M02300.XSHG 
            opt_type.append(ticker[6]) # opt_type记录'C'or'P'/C:认购 P认沽
            exp_date.append(ticker[7:11]) # exp_date 记录1503M数据中1503数据，1503合约到期日，M未发生除权，A调整一次B调整2次类推
        opt_data['optType'] = opt_type #追加dataframe 的columns 'optType'列认沽认购     'C'or'P'数据 
        opt_data['expDate'] = exp_date #追加'expDate'到期日列                          合约日期数据
        near_exp = np.sort(opt_data.expDate.unique())[0] #opt_data.expDate.unique()，去除到日期相同的数据，从小到大排列取第一个日期

   
        #计算所有上市期权：看涨看跌交易量、看涨看跌交易额、看涨看跌持仓量
        data = opt_data.groupby('optType') #将整个DataFrame按照'optType' 聚合，也就是C or P 进行分类
        hist_opt['callVol'][date] = data.turnoverVol.sum()['C'] #认购成交量总额
        #data聚合后turnoverVol的求和['C']分类的值附给hist_opt['callVol'][date]，date日期
        hist_opt['putVol'][date] = data.turnoverVol.sum()['P'] #认沽成交量总额
        hist_opt['callValue'][date] = data.turnoverValue.sum()['C'] #认购成交金额总额
        hist_opt['putValue'][date] = data.turnoverValue.sum()['P'] #认沽成交金额总额
        hist_opt['callOpenInt'][date] = data.openInt.sum()['C'] #认购持仓量总额
        hist_opt['putOpenInt'][date] = data.openInt.sum()['P'] #认沽持仓量总额
    
        # 计算近月期权(主力合约)： 看涨看跌交易量、看涨看跌交易额、看涨看跌持仓量    
        near_data = opt_data[opt_data.expDate == near_exp] #最近到期的合约数据
        near_data = near_data.groupby('optType') #最近到期合约数据按照'C'/'P'聚合分类
   
        hist_opt['nearCallVol'][date] = near_data.turnoverVol.sum()['C'] #近月认购成交量总额
        hist_opt['nearPutVol'][date] = near_data.turnoverVol.sum()['P'] #近月认沽成交量总额
        hist_opt['nearCallValue'][date] = near_data.turnoverValue.sum()['C'] #近月认购成交金额总额
        hist_opt['nearPutValue'][date] = near_data.turnoverValue.sum()['P'] #近月认沽成交金额总额
        hist_opt['nearCallOpenInt'][date] = near_data.openInt.sum()['C'] #近月认购持仓量总额
        hist_opt['nearPutOpenInt'][date] = near_data.openInt.sum()['P'] #近月认沽持仓量总额
        
        # 计算所有上市期权： 总交易量、总交易额、总持仓量
        hist_opt['netVol'][date] = hist_opt['callVol'][date] + hist_opt['putVol'][date] #认购+认沽成交量总额
        hist_opt['netValue'][date] = hist_opt['callValue'][date] + hist_opt['putValue'][date] #认购+认沽成交金额总额
        hist_opt['netOpenInt'][date] = hist_opt['callOpenInt'][date] + hist_opt['putOpenInt'][date] #认购+认沽持仓总额
        
        # 计算期权看跌看涨期权交易量(持仓量)的比率：
        # 交易量看跌看涨比率，交易额看跌看涨比率, 持仓量看跌看涨比率
        # 近月期权交易量看跌看涨比率，近月期权交易额看跌看涨比率, 近月期权持仓量看跌看涨比率
        # PCR = Put Call Ratio
        hist_opt['volPCR'][date] = round(hist_opt['putVol'][date]*1.0/hist_opt['callVol'][date], 4) #认沽成交量/认购成交量
        hist_opt['valuePCR'][date] = round(hist_opt['putValue'][date]*1.0/hist_opt['callValue'][date], 4) #认沽成交金额/认购成交金额
        hist_opt['openIntPCR'][date] = round(hist_opt['putOpenInt'][date]*1.0/hist_opt['callOpenInt'][date], 4) #认沽持仓量/认购持仓量
        hist_opt['nearVolPCR'][date] = round(hist_opt['nearPutVol'][date]*1.0/hist_opt['nearCallVol'][date], 4) #近月认沽成交量/认购成交量
        hist_opt['nearValuePCR'][date] = round(hist_opt['nearPutValue'][date]*1.0/hist_opt['nearCallValue'][date], 4) #近月认沽成交金额/认购成交金额
        hist_opt['nearOpenIntPCR'][date] = round(hist_opt['nearPutOpenInt'][date]*1.0/hist_opt['nearCallOpenInt'][date], 4) #近月认沽持仓量/认购持仓量    
    return hist_opt

def histPCR50ETF(beginDate, endDate):
    # PCRD: Put Call Ratio Diff
    # 计算每日PCR变化量：当日PCR减去前一日PCR得到的值，即对PCR做差分
    # 专注于某一项PCR，例如：成交额PCR --- valuePCR
    pcr_names = ['volPCR', 'valuePCR', 'openIntPCR', 
                 'nearVolPCR', 'nearValuePCR', 'nearOpenIntPCR']
    pcr_diff_names = [pcr + 'Diff' for pcr in pcr_names] #==['volPCRDiff', 'valuePCRDiff', 'openIntPCRDiff', 'nearVolPCRDiff', 'nearValuePCRDiff', 'nearOpenIntPCRDiff']
  
    pcr = histVolumeOpt50ETF(beginDate, endDate) #获取认沽认购期权数据
    for pcr_name in pcr_names:
       
        pcr[pcr_name + 'Diff'] = pcr[pcr_name].diff() #求一个差分
    return pcr[pcr_names + pcr_diff_names]
start = datetime(2015,2, 9)				# 回测起始时间
end  = datetime(2015, 10, 13)				# 回测结束时间

hist_pcrd = histPCR50ETF(start, end)      # 计算PCRD
# print hist_pcrd 
 
```
