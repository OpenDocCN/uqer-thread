获取日线波段

```python
df_index = DataAPI.MktIdxdGet(indexID=u"000001.ZICN",beginDate=u"19971201",field=u"tradeDate,closeIndex",pandas="1")  
df_index = df_index.set_index(['tradeDate'],drop=False)
piecewise_day = pieceWise(df_index.closeIndex.values)
piecewise_day.exceptionHandling()
```

指数没有周线数据，通过下面的函数可以将日线数据转化为周线数据

```python
def getWeeklyIndex(df_index):  # 得到周日期
    df_index['tradeDate'] = df_index['tradeDate'].map(lambda x:datetime.strptime(x ,'%Y-%m-%d'))
    df_index['year_week'] = df_index['tradeDate'].map(lambda x:datetime.isocalendar(x)[0:2])
    return df_index['tradeDate'].groupby(df_index['year_week']).tail(1)
```

获取周线波段

```python
df_week = getWeeklyIndex(df_index[:])
df_week_index = df_index.ix[df_week.index]['closeIndex']
piecewise_week = pieceWise(df_week_index.values)
piecewise_week.exceptionHandling()
```

自5178 点以来的下跌尚无法确认是否结束，因此暂时以2656 点为一个底部,加入到我们的拐点。

```python
piecewise_week.inflexion.append(list(df_week_index.index).index('2016-01-29')) #将'2016-01-29'加入到我们的拐点
```

# 上证综指2738是否为周线级别的底？
----
研报是5月初写的，站在当时的信息，得出这样的结论：`'如果从持续时间、对应段数这两个角度来看，我们认为此次周线级别下跌大概率没有结束。 目前处于第4段日线级别反弹中，反弹结束后， 2656必破。如果跌破 2656，走出 5 段下跌，这从段数上这轮周线级别下跌大概率结束。'`但现在的情况是(看下图)，第5段日线级别的下跌并没有破2656，第6段日线级别的反弹正在上演，目前指数正处于下跌趋势，但DEA还未小于零。

那么我个人的观点是，
* 如果7月15号至今这轮下跌破了2738（真实周线收盘价），那么从持续时间、对应段数这两个角度来看，都是OK的。持续时间从2015年6月12日到2016年8，9月左右，对应日线波段是7段，可以接受。那么这轮下跌大概率形成一个大底。

* 如果这轮下跌没有跌破2738，但形成了一个日线级别的底部，那么也有理由相信2738是周线级别的底。从对应段数来看，3段是OK的，下跌幅度47%也是OK的，唯一的问题是持续时间过短，但考虑到这轮下跌有自己的特点，融资爆仓、千股跌停、熔断等等，下跌比较快也是可以理解的。`倘若2738不是周线级别的底，那么这次周线级别的底至少要对应9段日线级别的波段，这个概率是比较小的。`

* 当然还有一个可能是这轮下跌只是轻微的调整，日线级别的上涨没有被打断。如果发生这种情况，我个人也是更偏向认为2738就是周线级别的底部。因为拿2004年4月2日到2005年6月3日周线级别下跌对应7个日线级别波段的情况来看，日线波段的底部是一直下移的，而如今日线级别的2016年5月19日的底部是大于2016年1月28日的底部，底部是上移。
