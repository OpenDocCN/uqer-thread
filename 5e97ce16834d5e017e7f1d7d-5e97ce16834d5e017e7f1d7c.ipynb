```python
# 转换stockcode代码为str类型
def code2str(x):
    x_str = str(x)
    while int(x)<100000:
        x = x*10
        x_str = '0' + x_str
    return x_str
```

```python
original_df = pd.read_csv('original_data.csv',encoding='GBK')
original_df['stockcode'] = original_df['stockcode'].apply(code2str)
```

# 设置基础参数
### 首先设置完基础参数之后就可以全部运行下面的脚本获取各个指标，会保存到相应的csv文档

```python
# 获取所有可转债，开始与结束日期
ticker = start_end_df['stockcode']
# 每次都str一下吧，确保是str
begin_date = '20100101'
end_date = '20191231'
begin_year = '2010'
end_year = '2019'
Q_basic_factor = ['ticker','secID', 'publishDate', 'endDate','reportType']
```

```python
# 因为经常用到上一期的值，排序+整体移动后要判断ticker是否相同
def same_ticker(df,str1='last'):
    if str1 == 'last':
        df['ticker1'] = ['0'] + df.ticker.tolist()[:-1]
    if str1 == 'next':
        df['ticker1'] = df.ticker.tolist()[1:] + ['0']
    return df.ticker == df.ticker1
# 因为要用到上一期的值，所以常常要排序
def sort_reset(df, sort_list):
    df.drop_duplicates(inplace = True)
    df.sort(sort_list,ascending=True, inplace = True)
    df.reset_index(drop = True, inplace = True)
    return df
```

```python
# 查看相同ticker,publishDate,endDate和reportType下，是否值是相同的。已经采用mergedFlag=='1'
def find_rep(df):
    def find_rep_element(x,df):
        df = df[df.ticker == x.ticker]
        df = df[df.publishDate == x.publishDate]
        df = df[df.endDate == x.endDate]
        df = df[df.reportType == x.reportType]
        if len(df)>1:
            return df
    re_df = pd.DataFrame()
    for index, row in df.iterrows():
        re_df=re_df.append(find_rep_element(row,df))
    return re_df


# 去掉原始数据中重复记录中endDateRep与endDate不同的记录
def delet_rep(df):
    def find_rep_element(x,df):
        df = df[df.ticker == x.ticker]
        df = df[df.publishDate == x.publishDate]
        df = df[df.endDate == x.endDate]
        df = df[df.reportType == x.reportType]
        if len(df)>1:
            # print('true')
            return True
        else:
            return False
    # 查找是否有相同的记录
    df['rep'] = df.apply(lambda x: find_rep_element(x,df),axis=1)
    # 查找endDate与endDateRep是否相同
    df['same_endDate'] = df.apply(lambda x: x.endDate == x.endDateRep,axis = 1)
    # 保留没有相同记录或者有相同记录，但是endDate一致的记录
    df = df[(df.rep==False)|(df.same_endDate==True)]
    return df
```

```python
# 根据给定本期，再判断上一期是否满足
# str为关键词
def fun2cond(df,str):
    df['last_cond'] = [np.NAN] + df.cond.tolist()[:-1] # 上一期是否满足条件
    df[str] = df.apply(lambda x: np.NAN if np.isnan(x.cond)|np.isnan(x.last_cond) else(x.cond&x.last_cond),axis = 1) # 本期上期均满足
    # 交接处的ticker不同造成的误判
    df['same_ticker'] = same_ticker(df,'last')
    df[str] = df.apply(lambda x: x[str] if x.same_ticker else(np.NAN),axis=1)
    return df
```

# 自建扣非因子数据
(新)合并利润表(所有会计期末最新披露) DataAPI.FdmtISLT2018Get

近2期年报扣非净利润均小于0或者本年度最新财报与上一年年报扣非净利润均小于0

```python
operateRato_factor = ['operateProfit','othOperCosts','investIncome','assetsImpairLoss','creditImpairLoss',
                      'netExpHIncome','fValueChgGain','assetsDispGain','forexGain']
```

```python
FdmtISLT2018Get = DataAPI.FdmtISLT2018Get(ticker=ticker,secID=u"",reportType=u"",
                                          beginDate=begin_date,endDate=end_date,year=u"",
                                          field=Q_basic_factor+operateRato_factor+['mergedFlag','endDateRep'],pandas="1")
new_df = FdmtISLT2018Get[FdmtISLT2018Get.mergedFlag == '1']
# new_df = delet_rep(df = new_df)
```

```python
# 构造扣非净利润
new_df.fillna(0,inplace=True)
new_df['operateRato'] = new_df[operateRato_factor].sum(axis = 1)
new_df['cond'] = new_df.operateRato.apply(lambda x: x<0)
new_df = new_df[Q_basic_factor+['cond']]
```

```python
# 条件1：近两期年报扣非净利润均小于0
def ifoperateRato_cond1(new_df):      
    # 提取是年报的部分，生成新的表
    A_df = new_df[new_df.reportType == 'A']     
    # 判断上期年报是否也满足条件
    A_df = sort_reset(A_df, ['ticker','publishDate','endDate'])
    A_df = fun2cond(A_df,'ifoperateRato_cond1')
    # 合并到原始表中
    new_df = pd.merge(new_df, A_df, left_on = new_df.columns.tolist(), right_on=new_df.columns.tolist(), how = 'left')
    return new_df.ifoperateRato_cond1.tolist()
```

```python
# 条件2：最新财报与上一年(本年指publishDate)年报(在本期财报发布日之前)扣非净利润均小于0
def ifoperateRato_cond2(x, new_df):
    if x.cond >= 0: # 本期不满足，则直接返回False
        return False
    else:
        last_year = str(int(x.publishDate[:4])-1)+'-12-31'
        temp_df = new_df[new_df.ticker == x.ticker]
        temp_df = temp_df[(temp_df.endDate == last_year) & (temp_df.reportType == 'A')]
        temp_df = temp_df[temp_df.publishDate <= x.publishDate]
        if len(temp_df)==0:
            return False
        else:
            temp_df.sort('publishDate',inplace=True,ascending=True) # 返回上年报告最新期的
            return temp_df.cond.tolist()[-1]
```

```python
new_df = sort_reset(df=new_df, sort_list=['ticker','publishDate','endDate'])
new_df['ifoperateRato_cond1'] = ifoperateRato_cond1(new_df)
new_df['ifoperateRato_cond2'] = new_df.apply(lambda x: ifoperateRato_cond2(x, new_df), axis = 1)
new_df['ifoperateRato_cond'] = new_df.apply(lambda x: (x.ifoperateRato_cond1 == True)|(x.ifoperateRato_cond2 == True),axis=1)
```

```python
# 返回有效值
new_df = new_df[Q_basic_factor+['ifoperateRato_cond']]
new_df = new_df[pd.isnull(new_df.ifoperateRato_cond) == False]
new_df.to_csv('ifoperateRato_cond.csv', encoding='GBK',index=False)
```

# ifoperatecash_cond
最近一期财报及上2期年报经营性现金流净额【和】小于0（除银行、非银金融、房地产）

```python
# NCFOperateA	float	经营活动产生的现金流量净额
FdmtCFGet = DataAPI.FdmtCFGet(ticker=ticker,secID=u"",reportType=u"",
                        endDate=end_date,beginDate=begin_date,publishDateEnd=u"",publishDateBegin=u"",endDateRep="",
                              beginDateRep="",beginYear="",endYear="",fiscalPeriod="",
                              field=Q_basic_factor+['NCFOperateA','mergedFlag','endDateRep'],pandas="1")
new_df = FdmtCFGet[FdmtCFGet.mergedFlag=='1']
new_df = delet_rep(new_df)
```

```python
# 是求和！！！不是对每一个数据进行判断！！！和上一个指标不一样
def A2_add(new_df):
    A_df = new_df[new_df.reportType == 'A'] 
    # 按照ticker排序，再按照publishDate升序排序
    A_df = sort_reset(A_df,['ticker','publishDate','endDate'])
    A_df['last_NCFOperateA'] = [np.NAN] + list(A_df.NCFOperateA.values[:-1]) # 本期对应的上一期
    A_df['same_ticker'] = same_ticker(A_df,'last') # 去掉交界处ticker不同的影响
    A_df['sum_2A'] = A_df[['last_NCFOperateA','NCFOperateA']].sum(axis = 1) # 本期和上一期求和
    A_df['sum_2A'] = A_df.apply(lambda x: x.sum_2A if x.same_ticker else(np.NAN),axis = 1) # 重新判断ticker交界处
    A_df = A_df[pd.isnull(A_df.sum_2A) == False]
    return A_df

# 对所有报告的publishDate进行排序，然后查找每一期财报最近的上一期年报，再把该期财报+上一期年报对应的两期年报值
def fun_operatecash_sum(x,new_df):
    sum_list = []
    for index,row in new_df.iterrows():
        while True:
            # 循环到，没有上期年报，则返回nan
            if new_df.same_ticker[index] == False:
                sum_list.append(np.NAN)
                break
            else:
                # 搜索到上期年报，返回和
                index -= 1
                if new_df.reportType[index]=='A':
                    sum_list.append(row.NCFOperateA + new_df.sum_2A[index])
                    break
    new_df['operatecash_sum'] = sum_list
    return new_df
```

```python
# 提取年报部分，通过排序将本期和上期年报求和
A_df = A2_add(new_df)
# 将最新两期年报的和条件合并到原始表中
new_df = pd.merge(new_df,A_df,left_on=list(new_df.columns),right_on=list(new_df.columns),how='left')
new_df = sort_reset(df=new_df, sort_list=['ticker','publishDate','endDate'])
new_df['same_ticker'] = same_ticker(new_df,str1='last')

# 将每一期财报与上两期年报求和
new_df = fun_operatecash_sum(new_df)
new_df['ifoperatecash_cond'] = new_df.operatecash_sum < 0
# 保存有效数据
new_df = new_df[Q_basic_factor+['ifoperatecash_cond']]
new_df = new_df[pd.isnull(new_df.ifoperatecash_cond)==False]
new_df.to_csv('ifoperatecash_cond.csv', index=False, encoding='GBK')
```

```python
# 指实征税额与实际收益额的比例。实征税额是应征税额扣除减免税额以后缴纳的税额。
FdmtIndiStctISGet = DataAPI.FdmtISGet(ticker=ticker,secID="",reportType=u"",
                                              field=Q_basic_factor+['incomeTax','TProfit','endDateRep','mergedFlag'],pandas="1")
FdmtIndiStctISGet = FdmtIndiStctISGet[FdmtIndiStctISGet.mergedFlag =='1']
FdmtIndiStctISGet = delet_rep(FdmtIndiStctISGet)
```

```python
def cal_itTP(x):
    if np.isnan(x.TProfit)|np.isnan(x.incomeTax):
        return np.NAN
    elif x.TProfit < 0.001:
        return np.NAN
    else:
        return x.incomeTax/x.TProfit

# 所得税/利润总额
FdmtIndiStctISGet['itTp'] = FdmtIndiStctISGet.apply(lambda x: cal_itTP(x), axis = 1)
# 每期是否满足条件
FdmtIndiStctISGet['cond'] = FdmtIndiStctISGet['itTp'] < 0.05

# 先进行排序，先按照ticker排序，再按照endDate排列，判断相邻两个是否为True，最后再恢复不同ticker交接处为np.nan
new_df = sort_reset(FdmtIndiStctISGet,['ticker','publishDate','endDate'])

# 判断上一期也是否满足
new_df = fun2cond(new_df,'iftax_cond')
# 提取有效值
new_df=new_df[Q_basic_factor+['iftax_cond']]
new_df = new_df[pd.isnull(new_df.iftax_cond)==False]
new_df.to_csv('iftax_cond.csv',encoding='GBK',index = False)
```

# ifholdRatio_cond	
最近一期财报及上一期年报机构持仓比例均低于5%（包括陆股通，信托公司除外）
按照endDate来group求和
按照endDate判断财报是否年报
这个publishDate可能不能去merge到季度因子中，因为披露时间是不同的
shName的长度大于3时，判断为机构股
holdPct 单位%

```python
def fun_ifholdRatio(new_df):
    # 搜索上一期年报
    cond_list = []
    for index,row in new_df.iterrows():
        while True:
            # 循环到，没有上期年报，则返回nan
            if new_df.same_ticker[index] == False:
                cond_list.append(np.NAN)
                break
            else:
                # 搜索到上期年报，返回
                index -= 1
                if new_df.reportType[index]=='A':
                    cond_list.append(new_df.cond[index])
                    break
    new_df['last_cond'] = cond_list
    return new_df
```

```python
EquShTenGet = DataAPI.EquShTenGet(secID=u"",ticker=ticker,beginDate=begin_date,endDate=end_date,
                                  field=['ticker','endDate','shName','holdPct','publishDate'],pandas="1")

# 判断是否为机构持股,汉字编码是三个为一个,提取四个字以上的
EquShTenGet['cond'] = EquShTenGet.shName.apply(lambda x: True if len(x)>3*4 else(False))
EquShTenGet = EquShTenGet.loc[EquShTenGet.cond == True]

# 判断是否为信托
EquShTenGet['cond'] = EquShTenGet.shName.apply(lambda x: True if '信托' in x else(False))
new_df = EquShTenGet.loc[EquShTenGet.cond == True]
```

```python
# 按照endDate和publishDate求和,用transform来扩展
new_df['sum'] = new_df.groupby(['ticker','endDate','publishDate'])['holdPct'].transform('sum')
# 删除由于publishDate为nan造成的holdPct的nan
new_df = new_df.loc[pd.isnull(new_df['sum'])==False]
# 保留有效的key
new_df.drop(['shName','holdPct'],axis = 1,inplace=True)
new_df = sort_reset(new_df,['ticker','publishDate','endDate'])

# 判断每一期是否满足条件
new_df['cond'] = new_df.apply(lambda x: x['sum'] < 5, axis =1)
new_df['same_ticker'] = same_ticker(new_df)   
new_df['reportType'] = new_df.apply(lambda x: 'A' if x.endDate[5:10]=='12-31'else(False),axis = 1)
# 上期年报是否满足条件
new_df = fun_ifholdRatio(new_df)
# 本期和上期年报是否都满足条件
new_df['ifholdRatio_cond'] = new_df.apply(lambda x: np.NAN if np.isnan(x.cond)|np.isnan(x.last_cond) else(x.cond&x.last_cond),axis = 1)

new_df = new_df[['ticker','endDate','ifholdRatio_cond']]
new_df = new_df[pd.isnull(new_df.ifholdRatio_cond)==False]
new_df.to_csv('ifholdRatio_cond.csv',encoding='gbk',index = False)
```

ROE_G_3y	因子profit_ttm（当日）除以tot_equity（最近一期已公布财报）三年复合增长率

```python
def cubic_root(nr):
    if nr<0:
        return -np.power(-nr, float(1)/3)
    else:
        return np.power(nr, float(1)/3)
```

```python
def cal_ttm(df,base_str,new_str):
    df[new_str] = df.apply(lambda x:df[(df.ticker==x.ticker)&(df.year==x.year-1)&(df.period==x.period)][base_str].values.tolist(),axis=1)
    df[new_str] = df[new_str].apply(lambda x:np.nan if len(x)==0 else x[-1])
    df[new_str] = ((df[base_str]-df[new_str])/df[new_str])
    return df

def cal_3y(df,base_str,new_str):
    df[new_str] = df.apply(lambda x:df[(df.ticker==x.ticker)&(df.year==x.year-3)&(df.period==x.period)][base_str].values.tolist(),axis=1)
    df[new_str] = df[new_str].apply(lambda x:np.nan if len(x)==0 else x[-1])
    df[new_str] = ((df[base_str]-df[new_str])/df[new_str]).apply(cubic_root)
    return df
```

# 合并季度因子

```python
# 首先根据债券存续期提取三大基础报表的有效数据，因为有的报表有一部分缺失，所以采用outer方法合并
# x是整体的值
# start_end_df 表示了开始和结束日期
def select_effective(x, start_end_df):
    temp1 = x.publishDate >= start_end_df.start_date
    temp2 = x.publishDate <= start_end_df.end_date
    temp3 = x.ticker == start_end_df.stockcode
    temp4 = start_end_df.loc[temp1 & temp2 & temp3]
    if len(temp4) == 0:
        return np.nan
    else:
        return temp4.ykcode.values[0]
def keep_effective_data(filename,start_end_df):
    temp_df = pd.read_csv(filename,encoding='gbk')    
    temp_df['ykcode']= temp_df.apply(lambda x: select_effective(x,start_end_df), axis = 1)
    temp_df = temp_df[pd.isnull(temp_df.ykcode)==False] # 仅保留存续期中
    return temp_df
```

```python
# 合并的时候重新读取数据，ticker为float形式，前面的0不要补齐，因为都没有
# 合并三大报表的基础季度因子
def basic_Q():
    start_end_df = pd.read_csv('start_end_df.csv',encoding='gbk')
    temp_df = keep_effective_data('FdmtIndiGrowthGet.csv',start_end_df)
    temp_df1 = keep_effective_data('FdmtIndiRtnGet.csv',start_end_df)
    temp_df2 = keep_effective_data('FdmtIndiTrnovrGet.csv',start_end_df)

    key_list = ['secID','publishDate','endDate','partyID','ticker','ykcode']
    
    new_df = pd.merge(temp_df,temp_df1,left_on = key_list, right_on = key_list,how = 'outer')
    new_df = pd.merge(new_df,temp_df2,left_on = key_list,right_on=key_list,how = 'outer')
    # 添加债券存续时间
    new_df = pd.merge(new_df,start_end_df,left_on = ['ykcode'],right_on=['ykcode'],how = 'left')
    new_df.reset_index(drop=True,inplace=True)
    new_df.drop(['ticker'],inplace=True,axis = 1)
    return new_df

new_df = basic_Q()
new_df.to_csv('Q_temp.csv',encoding='gbk',index=False)
```

# 合并行业指标和债券评级

```python
def add_industry(new_df,filename,ind_type):
    industry_df = pd.read_csv(filename,encoding='gbk')
    # industry_df 行业标签
    indID = []

    new_df.reset_index(drop=True,inplace=True)
    for i in new_df.index:
        temp1 = industry_df.ticker.values==new_df.stockcode.values[i]
        temp2 = industry_df.intoDate <= new_df.publishDate[i]
        temp3 = industry_df.outDate > new_df.publishDate[i]
        # 存在提取不到时间的
        temp4 = industry_df.industryID1[temp1&temp2&temp3]
        if len(temp4) == 0:
            indID.append(np.NAN)
        else:
            indID.append(temp4.values[0])
    new_df[ind_type] = indID
    return new_df

def add_bond(new_df,filename,ind_type):
    industry_df = pd.read_csv(filename,encoding='gbk')
    # industry_df 行业标签
    indID = []

    new_df.reset_index(drop=True,inplace=True)
    for i in new_df.index:
        temp1 = industry_df.secID.values==new_df.ykcode.values[i]
        temp2 = industry_df.intoDate <= new_df.publishDate[i]
        temp3 = industry_df.outDate > new_df.publishDate[i]
        # 存在提取不到时间的
        temp4 = industry_df.rating[temp1&temp2&temp3]
        if len(temp4) == 0:
            indID.append(np.NAN)
        else:
            indID.append(temp4.values[0])
    new_df[ind_type] = indID
    return new_df
```

```python
def fun_renew(x,list1,str1):
    if np.isnan(x[str1]): # 本身空值返回空值
        return np.nan
    elif x.industryZX1 in list1: # 属于特定各行业返回空值
        return np.NAN
    else: # 不属于特定行业，返回原值 
        return x[str1]
    
def fun_renew_com(x,str1):
    if np.isnan(x[str1]):
        return np.nan
    elif x.comp_type_wind == '国有企业':
        return np.NAN
    else:
        return x[str1]
```
