```python
import matplotlib.pyplot as plt
from matplotlib import rc
from matplotlib import dates
import seaborn as sns
import datetime
import numpy as np
import pandas as pd
import time
import scipy.stats as st
from CAL.PyCAL import * 
import lib.alpha as alp

cal = Calendar('China.SSE')
```

```python
start_date = '20090101'
end_date = '20180201'

universe = DynamicUniverse('A')

cal_dates = DataAPI.TradeCalGet(exchangeCD=u"XSHG", beginDate=start_date, endDate=end_date).sort('calendarDate')
cal_dates = cal_dates[cal_dates['isMonthEnd']==1]
trade_month_list = cal_dates['calendarDate'].values.tolist()

univ = []
for date in trade_month_list:
    univ = list(set(univ + universe.preview(date, skip_halted=True)))
```

```python
def const_trade_calendar():
    
    data = DataAPI.TradeCalGet(exchangeCD=u"XSHG,XSHE", beginDate=u"20050101", endDate=u"", field=['calendarDate', 'isOpen'])
    data = data[data.isOpen == 1].calendarDate.tolist()
    data = sorted(list(set(data))) 

    df = pd.DataFrame()
    df['trade_days'] = data

    return data, df


def next_day(day, n=1):
    
    _, df = const_trade_calendar() 
    df = df[df.trade_days > day]

    return df.head(n).trade_days.tolist()[n-1]


def get_factor(trade_date_list, universe):
    
    tra_year_list = pd.DataFrame(index=trade_date_list, columns=['tra_date'])
    for dt in trade_date_list:
        
        tra_year_list.ix[dt,'tra_date'] = next_day(str(dt), n=1)
    
    factor = pd.DataFrame()
    for dt in trade_date_list:
        
        veco = DataAPI.VecoRankCeGet(beginDate=dt, endDate=dt, field=["dataPeriod","rankObject","rankNO"], pandas="1")
        vecoID = DataAPI.EquGet(secID=universe, listStatusCD=u"L", field=["secID","secFullName"], pandas="1")

        cum_df = pd.merge(veco, vecoID, left_on='rankObject', right_on='secFullName', how='right')
        cum_df['dataPeriod'] = cum_df['dataPeriod'].fillna(method='ffill')
                           
        fac_df = cum_df.pivot(index='dataPeriod', columns='secID', values='rankNO')
        factor = factor.append(fac_df)
        
    fac = pd.merge(factor, tra_year_list, left_index=True, right_index=True, how='left')
    fac = fac.set_index('tra_date')
    
    return fac
```

```python
year_list = pd.date_range('2008-12-31', '2016-12-31', freq='12m')
fac = get_factor(year_list, univ)
fac.to_csv('veco.csv')
```
