```python
def cal_indicators(daily_ret, fre='daily'):
    if fre == 'daily':
        freq_ = 250
    if fre == 'monthly':
        freq_ = 12
    df_cum_value = (daily_ret + 1).cumprod()
    res = pd.DataFrame(index=['年化收益率','年化标准差','贝塔','夏普比率','信息比率','最大回撤'], 
                       columns=daily_ret.columns, data=0.0)
    
    res.loc['年化收益率'] = (daily_ret.mean() * freq_).apply(lambda x: '%.2f%%' % (x*100))
    res.loc['年化标准差'] = (daily_ret.std() * np.sqrt(freq_)).apply(lambda x: '%.2f%%' % (x*100))
    res.loc['贝塔', daily_ret.icol(1).name] = 1
    res.loc['贝塔', daily_ret.icol(0).name] = np.round((np.cov(daily_ret[daily_ret.icol(0).name],
                                             daily_ret[daily_ret.icol(1).name])[0,1] 
                                             / np.cov(daily_ret[daily_ret.icol(1).name])), 2)
    res.loc['夏普比率'] = (daily_ret.mean() / daily_ret.std() * np.sqrt(freq_)).apply(
                                                                        lambda x: np.round(x, 2))
    res.loc['信息比率', daily_ret.icol(0).name] = np.round(((daily_ret[daily_ret.icol(0).name].mean()
                                                         - daily_ret[daily_ret.icol(1).name].mean())  
                                                         / (daily_ret[daily_ret.icol(0).name]-
                                        daily_ret[daily_ret.icol(1).name]).std() * np.sqrt(freq_)), 2)
    res.loc['最大回撤'] = cal_maxdrawdown(df_cum_value).apply(lambda x: '%.2f%%' % (x*100))
    return res


def cal_maxdrawdown(data): 
    if isinstance(data, list):
        data = np.array(data)
    if isinstance(data, pd.Series):
        data = data.values    
    def get_mdd(values): 
        dd = [values[i:].min() / values[i] - 1 for i in range(len(values))]
        return abs(min(dd))
    if not isinstance(data, pd.DataFrame):
        return get_mdd(data)
    else:
        return data.apply(get_mdd)
```

```python
def get_factor(trade_date_list, universe, factor_name):
    factor = pd.DataFrame()
    for dt in trade_date_list:
        df_ST = DataAPI.SecSTGet(secID=universe, beginDate=dt, endDate=dt, field=['secID'])
        univ_ = [stk for stk in universe if stk not in list(df_ST['secID'])]
        dt_df = DataAPI.MktStockFactorsOneDayGet(tradeDate=dt, secID=univ_, field=['tradeDate', 'secID']+[factor_name])
        factor = factor.append(dt_df)
    factor = factor.pivot(index='tradeDate', columns='secID', values=factor_name)
    
    equ_info = DataAPI.EquGet(equTypeCD=u"A",secID=u"",ticker=u"",listStatusCD=u"",field=u"",pandas="1")
    equ_info = equ_info[['secID', 'listDate', 'delistDate']].set_index('secID')
    equ_info['delistDate'] = [x if type(x)==str else trade_date_list[-1] for x in equ_info['delistDate']]
    equ_info['delistDate'] = pd.to_datetime(equ_info['delistDate'], format='%Y-%m-%d')
    equ_info['delistDate'] = [x + timedelta(1) for x in equ_info['delistDate']]   
    equ_info['listDate'] = pd.to_datetime(equ_info['listDate'], format='%Y-%m-%d')
    equ_info['listDate'] = [x + timedelta(90) for x in equ_info['listDate']]
    
    factor.index = pd.to_datetime(factor.index)
    for sec in factor.columns:
        sec_info = equ_info.ix[sec]
        factor.loc[:sec_info['listDate'], sec] = np.NaN
        factor.loc[sec_info['delistDate']:, sec] = np.NaN
        
    factor.index = [date.strftime('%Y-%m-%d') for date in factor.index]
    factor.index.name = 'tradeDate'
    factor.columns.name = 'secID'
    return factor
```

```python
def st_remove(source_universe, date):
    df_ST = DataAPI.SecSTGet(secID=source_universe, beginDate=date, endDate=date, field=['secID'])
    return [s for s in source_universe if s not in list(df_ST['secID'])]

def new_remove(source_universe, date, day = 1):
    ticker = [stk[0:6] for stk in source_universe]
    period = '-' + str(day) + 'B'
    pastDate = cal.advanceDate(date, period)
    pastDate = pastDate.strftime("%Y-%m-%d")
    ipo_date = DataAPI.SecIDGet(partyID=u"",assetClass=u"e",ticker=ticker,
                                cnSpell=u"",field=u"ticker,listDate",pandas="1")
    remove_list = ipo_date[ipo_date['listDate'] > pastDate]['ticker'].tolist()
    ticker_list = [stk for stk in ticker if stk not in remove_list]
    secID_list = ticker2secID(ticker_list)
    return secID_list

def ticker2secID(ticker):
    universe = DataAPI.EquGet(equTypeCD=u"A",listStatusCD="L,S,DE,UN",field=u"ticker,secID",pandas="1") 
    universe = dict(universe.set_index('ticker')['secID'])
    if isinstance(ticker, list):
        res = []
        for i in ticker:
            if i in universe:
                res.append(universe[i])
            else:
                print i, ' 在universe中不存在，没有找到对应的secID！'
        return res
    else:
        raise ValueError('ticker should be list！')
```

```python
def excess_returns(fac, periods):
    n_quantile = 10
    cols_mean = [i+1 for i in range(n_quantile)]
    cols = cols_mean
    excess_returns_means = pd.DataFrame(index=fac.index, columns=cols)
    
    for dt in excess_returns_means.index:
        qt_mean_results = []
        tmp_illiq = fac.ix[dt].dropna()
        tmp_return = periods.ix[dt].dropna()
        tmp_return = tmp_return[tmp_return<0.6]
        tmp_return_mean = tmp_return.mean()
        pct_quantiles = 1.0/n_quantile
        for i in range(n_quantile):
            down = tmp_illiq.quantile(pct_quantiles*i)
            up = tmp_illiq.quantile(pct_quantiles*(i+1))
            i_quantile_index = tmp_illiq[(tmp_illiq<=up) & (tmp_illiq>=down)].index
            mean_tmp = tmp_return[i_quantile_index].mean() - tmp_return_mean
            qt_mean_results.append(mean_tmp)

        excess_returns_means.ix[dt] = qt_mean_results
    excess_returns_means.dropna(inplace=True)
    excess_returns_mean = excess_returns_means.mean()
    return excess_returns_mean
```

## <font color=#0099ff>**Smart Beta指数的基本表现**</font>
---
从指数表现上看，A股中大部分的Smart Beta指数均能稳定地获得超越基准指数的收益率

## <font color=#0099ff>**低波动风格的分析**</font>
---
高价值的长期风险溢价属性广为人知，所以我们来分析低波动的风险溢价来源

低波动风格的金融学解释 
根据传统的CAPM模型，当投资者投资于波动率高的股票组合时，投资者承担更高的市场风险时，相应地，投资者应当获得更高的超额回报。但上面的回测可以发现，投资于低波动的股票组合相比高波动组合反而能带来更高的回报，金融学中将这种情形称为低波动异象。 
低波动异象的几种解释 

1. `市场杠杆限制`: 风险偏好高的投资者倾向购买风 险高(波动大)的股票

2.  `博彩假说`: 投资者购买高波动股票期望以小博大

3. `分析师一致预期`: 分析师对高波动的股票的表现 过于乐观

4.  `跟踪误差理论`: 基金经理买入低波动股票以期获 得更稳健的回报和更低的跟踪误差 

为了进一步分析低波动风格因子的表现，我们进行分组研究，分别以沪深300和中证500为股票池构建低波动组合，观测`波动风格因子不同分组的表现`

## <font color=#0099ff>**低波动 + Smart Beta**</font> 
---
低波动风格相对其他风格表现出色，并且从风格暴露上看，它与几个比较有效的风格因子具有很强的相关性。从这个角度看，其他风格因子可以为低波动风格带来额外的超额回报，因此，我们可以主动地在低波动组合上增加其他风格因子的暴露，以提高策略的收益

基于这一想法，我们按照如下的策略构建方法来检验“低波动+Smart Beta”策略的有效性： 
1. 选定股票池（分别考虑中证800，中证500，沪深300三种情形），在股票池中剔除ST股, 60天新股
 
2. 按照单一因子的数据处理方法对股票进行去极端值，标准化，中性化

3. 月度调仓，按照 低波动和Smart Beta 因子筛选出排名前N个股票，构建股票组合。
 
综合考虑了策略的表现以及策略相比于原始策略（包括低波动组合以及原始因子组合）的表现的提升幅度，最终选择了以下Smart Beta策略组合：
1. 沪深300：低波动 + 高成长

2. 中证500：低波动 + 低估值

3. 中证800：低波动 + 高成长

结合 [【look 1 look 指数轮动的可能性】](https://uqer.datayes.com/v3/community/share/5baf9606b29fc401104cfd79), 以沪深300/中证500指数，构建月度的指数轮动模型

A.使用债券市场以及股票市场相关指标构建择时变量
- 利率类指标，波动率类指标，估值类指标在月频调仓下对于未来1个月沪深300与中证500指数收益差具有较强的预测效果 
- 指标分别是：期限利差变动，指数波动率截面差值，指数一致预期估值截面差值

B.使用Logistic回归进行指数相对强弱概率预测
- 使用Logistic回归滚动预测未来1期中指数收益差为正的概率
- 概率高于55%时，做中证500；概率低于45%时，做多沪深300；概率介于45%至55%之间时，做多中证800

**最终构建Smart Beta 的指数轮动策略**

```python
# 沪深300：低波动 + 高成长
HSIGMA_300 = 1 /get_factor(trade_month_end, univ_hs300, 'HSIGMA')
HSIGMA_300 = pretreat_factor(HSIGMA_300, neu=False)
OP_300 = get_factor(trade_month_end, univ_hs300, 'OperatingRevenueGrowRate')
OP_300 = pretreat_factor(OP_300, neu=False)
# 中证500：低波动 + 低估值(EP和CFP合成)
HSIGMA_500 = 1 /get_factor(trade_month_end, univ_zz500, 'HSIGMA')
HSIGMA_500 = pretreat_factor(HSIGMA_500, neu=False)
CFP = 1 /get_factor(trade_month_end, univ_zz500, 'PCF')
CFP = pretreat_factor(CFP, neu=True)
EP = 1 /get_factor(trade_month_end, univ_zz500, 'PE')
EP = pretreat_factor(EP, neu=True)
Value = (EP + CFP) /2
# 中证800：低波动 + 高成长
HSIGMA_800 = 1 /get_factor(trade_month_end, univ_zz800, 'HSIGMA')
HSIGMA_800 = pretreat_factor(HSIGMA_800, neu=False)
OP_800 = get_factor(trade_month_end, univ_zz800, 'OperatingRevenueGrowRate')
OP_800 = pretreat_factor(OP_800, neu=False)

fac_hs300 = HSIGMA_300 + OP_300
fac_zz500 = HSIGMA_500 + Value
fac_zz800 = HSIGMA_800 + OP_800
```
