# 海龟交易法则V2.0 系统I的突破附加处理

海龟交易法则  
系统I：中线策略是盘中股价突破20日最高价就买，跌破10日最低价就卖；  
系统II：长线策略是盘中股价突破55日最高价就买，跌破20日最低价就卖。  
另外有一个屏蔽器，当55日线在300日线下方的时候，意味这长线趋势不对，所以就算发出买入信号也不买。  
止损条件：买入价格跌破2N就快速止损，这个N跟TR有关。TR是真实波动幅度的缩写，公式是=max(H-L，H-PDC，PDC-L)：  
H=今日最高价，L=今日最低价，PDC=昨日收盘价  
3者取其大，N值就是过去20天的TR加起来除以20（均值）。
但是日线回测有比较严重的问题在于，这个海龟交易系统讲求盘中突破或者跌破，等到收盘突破并第二天开盘才动作的话，已经晚很多了，会造成偏差。  
V2.0来实现系统I的突破附加处理：  
1.上一次是盈利性突破，信号将被忽略（不管交易者当时是采取了突破交易，还是根据这一法则忽略了那次突破），直到如果能突破55日（保障性突破点）最高价入市，以避免错过重大趋势。  
2.如果上一次是亏损性突破（发生了2N幅度的不利变动），目前的新突破视为有效  

海龟交易法则还有很有技术含量的地方在于分散投资与关联品种控制，仓位控制、止损，这些都没有实现（另外最好要分时级别回测了，才能更精细控制）。  
取数部分写得比较复杂是因为要避免整个回测从300天后才开始。
