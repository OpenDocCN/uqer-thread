# 豆粕期货

## 豆粕期货 1，5，9合约基差套利策略研究 ##

豆粕由于大豆种植具有季节性的因素，其主力1，5，9合约所对应的现货成本和供给都有所不同，可以被看做不同品种的标的，然而由于最终产出品性质相同，因而其价格具有相当强的联动性，不同月份合约间的升贴水在正常情况下有相对较强的均值回归特性（我猜的），因而通过分析豆粕1，5，9合约之间的升贴水特性，研究升贴水的分布状况，可以就此研究出相对合理的1，5，9合约价差套利方案（包括期权）。

```python
## 豆粕期货 1，5，9 合约数据采选 ##
# 初始环境及信息准备 #

import pandas as pd 
import DataAPI as DA
import numpy as np
import matplotlib as mp
import datetime
import math
import sys

startdate='2011-01-01'
enddate='2018-04-23'
fee=6                     # 单边手续费
value=1000000
margin=0.05

m1=[]
m5=[]
m9=[]

# 准备合约代码 #
for i in range(1,10):
    m1.append('M'+str(i*100+1+1000))
    m5.append('M'+str(i*100+5+1000))
    m9.append('M'+str(i*100+9+1000))

# 提取交易日 #
tradedate=DA.TradeCalGet(exchangeCD=u"XDCE",beginDate=startdate,endDate=enddate,field=u"",pandas="1")
td= tradedate[tradedate['isOpen']==1]
td=td[u'calendarDate']
td=td.reset_index()[u'calendarDate']
dta1=pd.DataFrame([])
dta5=pd.DataFrame([])
dta9=pd.DataFrame([])
for j in range(1,10):
# 提取1月合约数据 #
    dt1=[dta1,pd.DataFrame(DataAPI.MktFutdVolGet(secID=u"",ticker=m1[j-1],beginDate=startdate,endDate=enddate,field=u"",pandas="1"))]
    dta1=pd.concat(dt1)
    dta1=dta1.reset_index(drop=True)
# 提取5月合约数据 #
    dt5=[dta5,pd.DataFrame(DataAPI.MktFutdVolGet(secID=u"",ticker=m5[j-1],beginDate=startdate,endDate=enddate,field=u"",pandas="1"))]
    dta5=pd.concat(dt5)
    dta5=dta5.reset_index(drop=True)
# 提取9月合约数据 #
    dt9=[dta9,pd.DataFrame(DataAPI.MktFutdVolGet(secID=u"",ticker=m9[j-1],beginDate=startdate,endDate=enddate,field=u"",pandas="1"))]
    dta9=pd.concat(dt9)
    dta9=dta9.reset_index(drop=True)
            
```

完成数据取值后，观察1，5，9合约间价差分布,并进行描述性统计

```python
# 以收盘价计算1，5，9合约的价差进行价差分析 #
# 1,5合约价差 #
dif15=np.abs(dta1[u'closePrice']-dta5[u'closePrice'])

# 5,9合约价差 #
dif59=np.abs(dta5[u'closePrice']-dta9[u'closePrice'])

# 9,1合约价差 #
dif91=np.abs(dta9[u'closePrice']-dta1[u'closePrice'])

plt.figure(1)
ax1=mp.pyplot.subplot(221)
ax5=mp.pyplot.subplot(222)
ax9=mp.pyplot.subplot(223)


plt.sca(ax1)
mp.pyplot.hist(dif15)
plt.sca(ax5)
mp.pyplot.hist(dif59)
plt.sca(ax9)
mp.pyplot.hist(dif91)

# 描述性统计 #
description15 = pd.DataFrame(index=['value'],columns=['percentile95','percentile80','mean','media', 'percentile20', 'percentile05','standard_deviation'])

description15['mean']=np.average(dif15)
description15['media']=np.median(dif15)
description15['percentile95']=np.percentile(dif15,95)
description15['percentile80']=np.percentile(dif15,80)
description15['percentile20']=np.percentile(dif15,20)
description15['percentile05']=np.percentile(dif15,5)
description15['standard_deviation']=np.std(dif15)


description59 = pd.DataFrame(index=['value'],columns=['percentile95','percentile80','mean','media', 'percentile20', 'percentile05','standard_deviation'])

description59['mean']=np.average(dif59)
description59['media']=np.median(dif59)
description59['percentile95']=np.percentile(dif59,95)
description59['percentile80']=np.percentile(dif59,80)
description59['percentile20']=np.percentile(dif59,20)
description59['percentile05']=np.percentile(dif59,5)
description59['standard_deviation']=np.std(dif59)


description91 = pd.DataFrame(index=['value'],columns=['percentile95','percentile80','mean','media', 'percentile20', 'percentile05','standard_deviation'])

description91['mean']=np.average(dif91)
description91['media']=np.median(dif91)
description91['percentile95']=np.percentile(dif91,95)
description91['percentile80']=np.percentile(dif91,80)
description91['percentile20']=np.percentile(dif91,20)
description91['percentile05']=np.percentile(dif91,5)
description91['standard_deviation']=np.std(dif91)


```
