# 第一题：按月计算最大召回率

```python
import pandas as pd
import pandas.io.data as web
import matplotlib.pyplot as plt
import datetime
import numpy as np
from dateutil.relativedelta import relativedelta

# maxinum.accumulate 当前坐标最大值
# 月最大回撤率计算函数
def get_max_draw_down(month_data):
    return ((np.maximum.accumulate(month_data) - month_data)/np.maximum.accumulate(month_data)).max()
    
    
#计算两个时间段相差多少个月的函数
def cal_months(date1,date2):
    year1=date1.year
    year2=date2.year
    month1=date1.month
    month2=date2.month
    num=(year1-year2)*12+(month1-month2)
    return num
        
    
#设置起始日期
start = datetime.datetime(2018, 1, 1)
end = datetime.datetime(2020, 1, 1)


# 获得平安银行数据：
out_data = DataAPI.MktEqudGet(secID=u"000001.XSHE", beginDate=start, endDate=end, pandas="1")
#转换成日期类型
out_data['tradeDate'] = pd.to_datetime(out_data['tradeDate'])
#由于需要按月取数据，设置日期为索引
out_data=out_data.set_index('tradeDate')
#只需要价格和日期数据，所以使用新数据
target_data=pd.Series(out_data['preClosePrice'],name='date',index=out_data.index)

#开始计算月最大撤回率
values=[]
dates=[]
now=start+relativedelta(months=-1)
for i in range(cal_months(end,start)):
    now=now+relativedelta(months=+1)
    query=now.strftime('%Y-%m')
    dates.append(query)
    month_data=list(target_data[query])
    values.append(get_max_draw_down(month_data))

#展示数据
plt.figure(figsize=(12,4))
plt.xticks([i for i in range(len(dates))],dates,rotation=30)
plt.title('PingAn Analysis')
plt.xlabel("month")
plt.ylabel("max draw down")
plt.plot([i for i in range(len(dates))],values)

plt.show()
```
