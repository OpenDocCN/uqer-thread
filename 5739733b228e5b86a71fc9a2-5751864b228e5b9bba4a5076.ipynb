今年3月9日，阿尔法围棋以4：1的成绩战胜了职业九段选手李世石。阿尔法围棋利用“价值网络”计算局面，用“策略网络”选择下子。阿尔法围棋的一个神经网络大脑是“监督学习的策略网络{Policy Network}”，观察棋盘布局的最佳下一步。计算机在面对重压下，不会受到情绪左右。阿尔法围棋的原理一样可以运用到资管业。
 
一直以来，信息科技驱动投资革命。投资最核心的有三个主要的要素。第一是专业领域的知识。这是长时间逐步积累起来的，很多的专家、学者帮我们积累专业知识。第二是获取信息的能力。如果所有人都使用同样的信息和方法，那么投资的结果或决策基本没什么太大差异。如果我们获取了与别人不一样的信息或数据，就可以建立起自己的比较优势。第三是发现投资机会的能力。仅拥有同样的信息和数据，还是很难与竞争对手去竞争，而分析工具在很大程度上就可以帮助我们提升投资的能力。

**投资研究的发展过程经历了3个阶段。在这个过程中，技术对整个投资的过程和影响非常重要。**

计算机未出现时，投资者（主要指研究员、基金经理）通过看纸质报告，凭投资逻辑的理解进行判断，用纸笔进行研究，然后通过会议进行信息沟通和交流。这种情况下，效率低下、信息量极少，可储存信息有限，知识的重要性占比很大。

随着计算机的出现和发展，投资者的数据收集和处理能力加强，他们开始依赖计算机帮助自己提升投资研究能力。将原来的投资逻辑通过构建一些模型，就可以借助计算机进行分析和研究。

如今，我们进入了真正的移动互联网时代，数据采集和处理能力大大提升，云计算能够做非常复杂的模拟。以前我们靠专业知识构建模型，现在我们通过机器来学习，找到最核心的要素。构建机器学习的过程就是我们构建智能化投资平台的过程。

                                          投资管理的牛顿定律

有几种传统投资为人熟知。第一是基本面投资，像巴菲特等都是用非常传统的基本面分析进行投资。第二是技术分析，通过价格和交易量等市场信息发现投资机会。还有主题投资、事件投资等，近几年发展也比较快。**基本面投资的框架主要有这三个阶段：1.理解过去、2.预测未来、3.如何建立一个模型来估值和决策。 **

2013年诺贝尔经济学奖的获得者Robert Shiller，提出了一个更加完整的公式，引入了投资者情绪对于价格的影响。投资者情绪越高，价格就越偏离基本价值。在这个公式中有一点需要关注：套利成本。如果套利成本低，就会有人迅速纠正市场情绪造成的偏离，那么投资者情绪对价格的影响就会很小。反之，如果套利成本很高，就像现在，期货做空，空方的α就会比多方的α好很多。这个公式有点像爱因斯坦的相对论对牛顿定律的修正，更加系统的把投资的大概框架建立得更加完善。**在研究中，我们不仅要关注基本面，还要关注市场的行为、市场的情绪对投资的影响。**

进行基本面研究可以完全从市场的角度来分析。我们假设市场有所有的信息，这样所有的信息都包含在市场里，因为市场反映了基本面和市场情绪。同时，我们也假设市场是有一定的规律的，因为有一定规律我们才能找到一个模型或一种方法，从而利用这个规律进行投资。我们在研究的时候，首先要观察数据，发现里面的规律，然后确定买卖的时间点。所有的技术分析人员基本都遵循这个逻辑。他们通过分析很多观察到的现象，确定一些方法论。这种投资的特点是技术分析人员只关注价格或者价量的变化，通常用图表来分析它们的趋势和周期性。这种操作一般都是相对短期的，很少有人用这种投资方法做像巴菲特一样非常长期的三年、五年的投资。**在投资学中，我们把巴菲特深入研究股票基本面，少而精准地分析股票的偏艺术风格投资称为主动投资。另一种投资就是量化大师西蒙斯倡导的量化投资，通过广度地挖掘市场规律进行投资，偏科学性。**

> **投资已经从艺术发展为一门科学**

我先对量化投资做一点简单介绍。定性投资和量化投资，没有本质的区别，基本上都使用传统的投资方法、模型。基本面投资的一些理论，基本都可以用定量的方法表述出来并得以实现。无论是基本面还是技术面的理论，都可以用模型来表述。定性投资偏艺术，比如判断一家公司运营情况是好是坏，每位投资者的判断是不同的。如果想把某家公司研究透彻，需要花费很多精力，因此使用定性投资的话，投资者不可能覆盖很多公司。只能筛选出自己认为值得深入研究的几家公司，对其进行研究。不同于定性投资，定量投资采取大数法则。我们建立一个模型，可以应用到很多样本上，而且整个风险的分散比传统投资要好。虽然每一项投资赚的钱可能不是很多，但是覆盖面很广。这就是定性投资和量化投资主要的差异。

接下来我们一起来看量化投资的大框架，基本分为这样几部分：

第一，收益预测是核心。做量化投资，最核心、最基本的要素是寻找到阿尔法。无论是用估值分析、情绪分析还是市场分析，都要尽可能找到一个预测阿尔法的方法。

第二，风险管理是重要部分，它是做量化非常重要的环节。

第三，费用的管理不可忽视。很多人经常会发现自己做了一个很好的策略，策略跑的也很好，可是一旦真正操作就会发现，本来可以赚100%的收益，却亏钱了。这就是因为没有管理好费用。如果每个交易的阿尔法不能覆盖自己的成本，基本就是亏钱。

**这三点是比较核心的部分。**

第四，绩效分析或归因分析也很重要。当我们做很多投资的时候，就需要一套系统的监控系统去监控哪些策略跑得好，哪些跑得差。因为每个策略不可能永远都是好的，随着发现这个策略的人越来越多，效果就会越来越差。这时就要监控市场上哪些策略已经被很多人发现，哪些依旧拥有自己独有的特性。

**这四点是量化投资的基础内容。**

> **大数据在金融行业中的应用**

一个简单的股票定价模型包括三个因素：现金流、折现率和市场情绪(相当于百度指数，测度情绪指标)。传统方法可能用企业的财报，卖方的研究报告去作出预测，这些预测都像盲人摸象，看到的都是事物的某一个角度;但是在有了大数据之后，数据的维度越来越多，数据的及时性越来越强，使得投资者更容易看到事物的全貌;综合起来，我们的投资更趋智能化，更加依靠模型和数据去寻找规律，这些 是对于传统投资的一些改变。

大数据也使得量化投资从手工建模的阶段逐步朝机器帮我们找规律、更加智能投资的方向去发展。
