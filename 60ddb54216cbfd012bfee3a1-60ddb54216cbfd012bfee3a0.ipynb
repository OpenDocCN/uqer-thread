```python
import numpy as np
import pandas as pd
from CAL.PyCAL import *
cal= Calendar('China.SSE')

tickers_columns=['ticker','close_price','discountRatio0','discountRatio1','inc_percent','volatility' ,'secShortName']

pd.set_option('display.height',1000)
pd.set_option('display.max_rows',500)
pd.set_option('display.max_columns',500)
pd.set_option('display.width',1000)

def get_tickers_etf(trade_date):
    if cal.isHoliday(trade_date):
        trade_date= cal.advanceDate(trade_date, '-1B', BizDayConvention.Following).strftime("%Y%m%d")
    tickers_data = DataAPI.FundGet(secID=u"",ticker=u"",etfLof=u"ETF",listStatusCd=u"L",category=u"E",\
    idxID=u"",idxTicker=u"",operationMode=u"",beginDate="",endDate="",status="", \
                                   field=u"ticker,secID,idxID,listDate,idxTicker,secShortName,category,isQdii",pandas="1")
    tickers_data.sort_values(by='listDate', inplace=True);
    tickers_data.drop_duplicates(subset=['idxID'], keep='first', inplace=True)
    tickers_list = (tickers_data[tickers_data.isQdii != True])['ticker'].tolist()

    tickers_str = ",".join(tickers_list)
    tickers_data=DataAPI.MktFunddGet(secID=u"",ticker=tickers_str,tradeDate=trade_date,beginDate="",endDate="",field=u"ticker,secID,turnoverValue,secShortName",pandas="1")
    tickers_data=tickers_data[tickers_data.turnoverValue> 10000000.0]
    tickers_list = tickers_data['ticker'].tolist()
    return  tickers_list

def get_tickers_close(trade_date):
    if cal.isHoliday(trade_date):
        trade_date= cal.advanceDate(trade_date, '1B', BizDayConvention.Following)
   #获取交易日基金列表
    tickers_data=DataAPI.MktFunddGet(secID=u"",ticker=u"",tradeDate=trade_date,beginDate="",endDate="",field=u"ticker,discountRatio,turnoverValue,secShortName",pandas="1")
   # 过滤掉折价率 小于3% 和成交量 小于10000的
    filterData= tickers_data[(tickers_data.discountRatio > 0.03) & (tickers_data.turnoverValue> 10000.0)]
    tickers_list = filterData['ticker'].tolist()
    tickers_str = ",".join(tickers_list);
    tickers_data = DataAPI.FundGet(secID=u"",ticker=tickers_str,etfLof=u"",listStatusCd=u"",category=u"",\
    idxID=u"",idxTicker=u"",operationMode=u"",beginDate="",endDate="",status="", \
                                   field=u"ticker,secID,idxTicker,secShortName,category,isQdii,isClass",pandas="1")
   #  过滤 分级基金
    tickers_data.drop_duplicates(subset=['ticker'], keep='first', inplace=True)
    tickers_list = (tickers_data[tickers_data.isClass == 0])['ticker'].tolist() 
    return  tickers_list

def display_tickers(trade_date,tickers_list):
    if cal.isHoliday(trade_date):
        trade_date= cal.advanceDate(trade_date, '-1B', BizDayConvention.Following).strftime("%Y%m%d")
        
    print "cur date is {0}".format(trade_date);
    tickers_str = ",".join(tickers_list)
    tickers_data=DataAPI.MktFunddGet(secID='',ticker=tickers_list,tradeDate=trade_date,\
                                         beginDate="",endDate="",\
                                     field=u"ticker,secID,closePrice,CHGPct,discountRatio,turnoverValue,secShortName",pandas="1")
   # filterData= tickers_data[(tickers_data.discountRatio > 0.04) & (tickers_data.turnoverValue> 500000.0)]
    filterData =tickers_data.sort_values(by='discountRatio',ascending=False)
    print filterData

today = Date.todaysDate()

today_str= today.strftime("%Y%m%d");
yesterday_str = cal.advanceDate(today_str, '-1B', BizDayConvention.Following).strftime("%Y%m%d")

tickers_list = get_tickers_close(yesterday_str)
#print(tickers_list)
display_tickers(yesterday_str,tickers_list);
```
