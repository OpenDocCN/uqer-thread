# 申万银行板块配对

```python
import numpy as np
import pandas as pd
from pandas import *
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.stattools import coint

#一、先求银行股的协整关系

#1.先求申万一级行业银行板块的成分股
universe=set_universe(IndSW.YinHangL1,'20180101')
print universe
stocks=[] #这个列表用于装通过单位根检验的成分股
#2.确定时间周期，获取期间前复权数据，成分股收盘价一阶差分，删除空值;
start='20180101'
end='20180630'
for stock in universe:
    CPSeries=DataAPI.MktEqudAdjGet(secID=stock,beginDate=start,endDate=end,field='closePrice',pandas="1")
    diff_CPSeries=CPSeries.diff(1)
    #要注意，statsmodel的api不允许有na值出现，所以要dropna
    diff_CPSeries.dropna(inplace=True)
#diff_CPSeries本来时[[1],[2],[3],...]，加.T[0]变为[1,2,3,...]
    diff_CPSeries=np.array(diff_CPSeries).T[0]
#3.做单位根检验,选出p-value小于一定阈值的成分股
    try:
        zz=adfuller(diff_CPSeries)
    except:
        print stock,len(diff_CPSeries)
    #zz[0]是t统计值，zz[1]是t值对应的p-value，zz[2]是lag,zz[3]是样本数，zz[4]是供参考的显著性水平及相对应的p-value
    if zz[1]<0.005:
        stocks.append(stock)
print stocks


```

```python
#银行股中共有22只股票一阶差分的数据平稳
print len(stocks)
```

```python
#4.计算协整关系，2018年上半年，共有34对股票很可能存在协整关系
pair_trade=[]
for i in range(len(stocks)):
    for j in range(i+1,len(stocks)):
        data1=DataAPI.MktEqudAdjGet(secID=stocks[i],beginDate=start,endDate=end,field="closePrice",pandas="1")
        data2=DataAPI.MktEqudAdjGet(secID=stocks[j],beginDate=start,endDate=end,field="closePrice",pandas="1")
        x=np.array(data1).T[0]
        y=np.array(data2).T[0]
        try:
            t,pvalue,d=coint(x,y)
        except:
            pass
        if pvalue<0.05:
            pair_trade.append([stocks[i],stocks[j]])

```

```python
df_pair=pd.DataFrame(pair_trade,index=range(0,len(pair_trade)),columns=['name1','name2'])
df_pair
```
