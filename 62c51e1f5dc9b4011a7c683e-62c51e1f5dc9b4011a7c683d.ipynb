```python
# 获取某日可转债指数所有成分
def get_bond(tradeDate):
    bond=DataAPI.IdxConsGet(ticker=u"000832",intoDate=tradeDate,field=u"consID",pandas="1")['consID']
    tmp_ID=list(bond.values)
    #去掉可交换债
    cb=DataAPI.BondGet(secID=tmp_ID,typeID=u"02020113",field=u"secID",pandas="1")
    bond=list(cb['secID'])#获取每月最后一个交易日
    return list(bond) 
cal=DataAPI.TradeCalGet(exchangeCD=u"XSHG",beginDate=u"20180401",endDate=u"20220701",isOpen=u"1",field=u"",pandas="1")
cal = cal.loc[cal['isMonthEnd']==1]
cal.index = range(len(cal))
print(len(cal))
result={}
to_hold=[]
for i in range(len(cal)):
    if i!=0:
        #获取本月收益
        if len(to_hold)>0:
            price=DataAPI.MktBonddGet(secID=to_hold,beginDate=cal.loc[i-1]['calendarDate'],endDate=cal.loc[i]['calendarDate'],
                                      field=u"secID,tradeDate,openPrice,closePrice",pandas="1")
            price=price.groupby(['secID','tradeDate']).mean().unstack().T
            openPrice=price.loc['openPrice'].iloc[0]
            closePrice=price.loc['closePrice'].iloc[-2]#调仓日前一天卖出
            #print openPrice,closePrice
            mean = (closePrice/openPrice-1).mean()
            if 2 > mean > -1:
                mean = mean
            else:
                mean = 0
            result[cal.loc[i]['calendarDate']] = mean    
        else:
            result[cal.loc[i]['calendarDate']]=0
    bond=get_bond(cal.loc[i]['calendarDate'])#获取所有可转债
    #获取可转债
    z = DataAPI.BondRatingGet(secID=bond, pandas="1")
    z.drop_duplicates(subset="secID", keep="last", inplace=True)
    z = z.loc[z['rating']=='AA-']
    to_hold = list(z['secID']) 
    print(cal.loc[i]['calendarDate'],len(to_hold),to_hold[:5])
print('end')
    
```

```python
import pandas as pd
(pd.Series(result)+1).cumprod().plot()
print(pd.Series(result))
```
