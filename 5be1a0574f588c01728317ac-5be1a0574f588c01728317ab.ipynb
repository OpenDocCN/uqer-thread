# matplotlib练习

```python
#导入数据包
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#柱状图，折线图，散点图的画法
fig = plt.figure(figsize = (20,10))
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
years = [1950,1960,1970,1980,1990,2000]
gdp = [19500,19600,19700,19900,20100,20000]
ax1.bar(years,gdp)
ax2.plot(years,gdp)
ax3.scatter(years,gdp)
```

```python
#将数据的标签附上及数据居中
data = [42,63,99,18,69]
label = ['A','B','C','D','E']
plt.xticks(range(len(data)),data)
plt.xlabel('Class')
plt.ylabel('Amount')

plt.bar(range(len(data)),data,align="center")
for x,y in zip(range(len(data)),data):
    plt.text(x, y,y,horizontalalignment="center",verticalalignment="bottom")
plt.show()
```

```python
#柱状图的堆积
data1 = [66,59,83,169,85]
data2 = [54,48,65,94,74]
plt.bar(range(len(data1)),data1,width = 0.5,color = 'green')
plt.bar(range(len(data2)),data2,width = 0.5,color = 'navy',alpha = 0.9,bottom=data2)#堆积体现在bottom
```

```python
#柱状图的分条
data1 = [66,59,83,169,85]
data2 = [54,48,65,94,74]
width = 0.3
plt.bar(np.arange(len(data1)),data1,width = 0.3,color = 'yellow')#注意这里不要用range(len(data1)),因为列表不能与数字直接相加
plt.bar(np.arange(len(data2))+width,data2,width = 0.3,color = 'blue')#分条体现在"+width"
plt.show()
```

```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
ax1 = plt.subplot(2,1,1)
ax2 = plt.subplot(2,1,2)
x = np.linspace(0,10,100)
for i in range(5):
    plt.figure(1)
    plt.plot(x,np.exp(i*x/3))
    plt.sca(ax1)
    plt.plot(x,np.sin(i*x))
    plt.sca(ax2)
    plt.plot(x,np.cos(i*x))
    
plt.show()
```
