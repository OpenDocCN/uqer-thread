```python
#来自量化学堂，第1课 Python入门及数据处理

#PE是指股票的本益比，也称为“利润收益率”。本益比是某种股票普通股每股市价与每股盈利的比率。所以它也称为股价收益比率或市价盈利比率(市盈率）。此解释来自百度词条
#获取20180510沪深300股票当前的行业PE值，并进行排序
#1、获取所有股票的PE
#2、010303-申万行业，获取每只股票的行业信息
#3、沪深300成分股信息

trade_Date = "20180521"
df1 = DataAPI.MktEqudGet(tradeDate=trade_Date,secID=u"",ticker=u"",beginDate=u"",endDate=u"",isOpen="",field=u"",pandas="1")
#DataAPI.MktEqudGet,获取沪深AB股20180510日行情信息
df1 = df1[['secID', 'ticker', 'PE']]
#取出 证券编码、股票交易代码和PE 三列信息

df2 = DataAPI.EquIndustryGet(industryVersionCD=u"010303",industry=u"",secID=u"",ticker=u"",intoDate=u"",field=u"",pandas="1")
#DataAPI.EquIndustryGet,输入证券ID或股票交易代码，获取股票所属行业分类，industryVersionCD，行业分类标准数字编码。010301-证监会行业V2012、010303-申万行业、010314-中证行业
df2 = df2[df2['isNew']==1][['secID', 'ticker', 'industryName1', 'industryName2', 'industryName3']]
#取出最新的行业分类的'secID', 'ticker', 'industryName1', 'industryName2', 'industryName3'五列信息

df3 = DataAPI.IdxConsGet(secID=u"",ticker=u"000300",intoDate=u"",isNew=u"",field=u"",pandas="1")
#DataAPI.IdxConsGet,获取国内外指数的成分构成情况，包括指数成分股名称、成分股代码、入选日期、剔除日期等
df3 = df3[df3['isNew']==1][['consID', 'consTickerSymbol']]
#取出最新的 成分股内部编码、成分股代码 信息


#重新建索引
df1 = df1.set_index('secID', drop=True)
df2 = df2.set_index('secID', drop=True)
#把证券编码作为索引


#构造沪深300的成分股内部编码的列表
#得到hs300的成分股
#print df3['consID']
hs300 = df3['consID'].tolist()

#得到沪深300成分股的ticker和PE的DataFrame
df1 = df1.ix[hs300]
#loc——通过行标签索引行数据 
#iloc——通过行号索引行数据 
#ix——通过行标签或者行号索引行数据（基于loc和iloc 的混合） 
#同理，索引列数据也是如此！


#将两个dataframe合并
df_merge = pd.merge(df1, df2, how='left', on='ticker')
#把沪深300成分股的行业信息与PE合并

df_merge.groupby('industryName1')
#沪深300的成分股按第一行业进行分类

#统计行业均值
df_group = df_merge.groupby('industryName1').agg({'PE':'mean'}) 
#求出沪深300里每一类行业的平均PE

#排序
df_group = df_group.sort_values('PE', ascending=False)
df_group
#行业PE降序排序
```
