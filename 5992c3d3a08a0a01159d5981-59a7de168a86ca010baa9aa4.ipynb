####AH股交易策略
---

####策略思路

	对于同时在大陆和香港上市的公司，由于同股同权，同一公司的AH股的股票价格本应相同。但由于两地利率，税率以及政策和汇率等宏观因素影响，交叉上市的公司的股价在A股和H股会有一定的溢价。但溢价会保持相对稳定。
	在此基础上，可以通过对近期AH股价做线性回归，得到一个AH股价关系，由此关系，可以在当前H股价格情况下求得A股价格期望值。当A股实际价格低于A股价格的期望值时，在认为H股相对理性的前提下，A股可能会被低估。此时，会因为AH价格关系回归稳定而有获利空间。根据这个原理，H股可以认为是A股价格的风向标。
    
####策略实现

	1.首先，我们要判断AH股近期价格存在稳定的回归关系，这要求AH股价关系是强相关关系。
    2.对近期AH股价做线性回归，找到AH价格线性关系。
    3.判断，当A股价格低于A股价格期望值的0.5倍A股价格标准差时，认为A股相对被低估作为买入点的判断条件。
    4.加强判断，有时，AH股同时在下跌，但仍存在A股低于期望值的情况。为确定A股位于上涨趋势，加入A股五日平均大于十日平均条件

首先要在AH交叉上市的股票中选择具有强相关的股票，为简单起见我们选取2对已经验证具有强相关关系的AH股票对。在只能操作A股的情况下，股票池中有2只股票，假设取为 600188，601727 两只股票，以600188为例，看一下AH股走势关系

```python
import matplotlib.pyplot as plt
StockData = DataAPI.MktAHCompdGet(ticker=u"600188", beginDate=u"2016-06-01", endDate=u"2017-07-01")

plt.figure(1)
StockData = StockData.set_index("tradeDate")
StockData["closePriceA"].plot(label = "A Price")
StockData['closePriceH'].plot(label = "H Price")
plt.legend(loc='best')
plt.ylim(0,20)
plt.ylabel("Price")
```

虽然股票价格不一样，但他们始终保持相同趋势。但股票对价格偶尔会有偏离情况发生。
