# 筹码分布简易算法

```python
# 成交量的平均价格分布（筹码分布简易算法）
import numpy as np
import pandas as pd

strBegin = '20190501' #起始时间
strEnd = '20190712'   #该日的筹码分布
ticker = '300119'       #股票代码

# 0到100元分成10000分
priceVol = np.zeros( 10000 )

# 获取历史数据
data = DataAPI.MktEqudAdjGet(secID=u"",ticker=ticker,tradeDate='',beginDate=strBegin,endDate=strEnd,isOpen="1",field=u"secID,secShortName,tradeDate,closePrice,lowestPrice,highestPrice,turnoverVol,turnoverRate",pandas="1")

# 计算当日价格区间每分钱的成交量。简单平均法
data['volPerCent'] = data['turnoverVol'] / ((data['highestPrice'] - data['lowestPrice']) * 100)

for idx in data.index:
    # 该日的价格区间
    low = int(round(data['lowestPrice'][idx] * 100))
    high = int(round(data['highestPrice'][idx] * 100))

    # 整个价格空间每一价格减去该价格的成交量。筹码量=当前筹码量*(1-换手率)
    priceVol *= 1 - data['turnoverRate'][idx]
    
    # 当日价格区间的每一价格加上该价格的成交量
    priceVol[low:high] += data['volPerCent'][idx]

# 输出结果
res = pd.DataFrame(priceVol[0:2500])
res.plot()
```
