# 灵狐4号

多因子SVM结合逻辑回归三日选股策略（灵狐4号）

总体思路：单日上涨下跌突变性较大，因此重点关注训练三日涨幅超过2%的A股，并进行预测。
1、选出T-3交易日的A股所有股票的22个因子数值作为训练集。
2、选出T-2、T-1、T交易日的所有A股三日总计收益率作为训练标签结果，大于2%标记1，小于2%标记0。
3、选出T日的A股所有股票的22个因子数值作为测试集。
4、数据处理，包括因子数值标准化、停牌及非交易日的空值处理、涨停股票的去除。
5、用SVM算法对1和2训练，得到模型clf。
6、再用3的数据输入到clf模型中，得到明日的所有股票预测标签为1的股票作为下一步的股票池。
7、取第6步用SVM算法预测上涨的股票作为股票池，再用逻辑回归算法算这些股票的上涨概率，概率最大的20只股票作为持仓，在尾盘买入。

选出的22个因子见下方：（包含技术因子和基本面因子）
PB 市净率
PE 市盈率
PS 市销率
LCAP 对数市值
CTOP 现金流市值比
DebtEquityRatio 产权比率
ROE 权益回报率
ROA 资产回报率
EPS 每股收益
CurrentRatio 流动比率
DAVOL10 10日平均换手率与120日平均换手率（Turnover Rate）之比
DAVOL5 5日平均换手率与120日平均换手率（Turnover Rate）之比
MA5 5日移动平均
MA10 10日移动平均
MA20 20日移动平均
MFI 资金流量指标
ATR6 6日均幅指标
BIAS10 10日乖离率
CCI5 5日顺势指标（动量因子）
BollDown 下轨线（布林线）指标
BollUp 上轨线（布林线）指标
ROC6 6日变动速率（动量因子）
