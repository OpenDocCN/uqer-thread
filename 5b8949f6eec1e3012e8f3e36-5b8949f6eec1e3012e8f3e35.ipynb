# 历史市盈率回顾

我们可以利用指数走势与指数市盈率、市净率因子的相关性，定性地判断当前市场位置。

```python
# 初始化 
import pandas as pd
import matplotlib.pyplot as plt

# 获取指数日线行情信息，包含昨收价、开盘价、最高价、最低价、收盘价、成交量、成交金额等字段
# ticker={'000001':'上证综指'，'000009'：'上证380'，'000300':'沪深300'，'399001':'深证成指'，'399006':'创业板指'} ，修改指数代码ticker可以查看各指数的数据
ticker=u'000001'
df1=DataAPI.MktIdxdGet(indexID=u"",ticker=ticker,beginDate=u"20070101",endDate=u"",exchangeCD=u"XSHE,XSHG",field=u"tradeDate,closeIndex",pandas="1")

# 获取指数估值信息，包含市值、PE、PB、ROE等信息，目前涵盖46个指数，每日18:30更新。 上证综指的最早数据是20180505日的数据 
df2=DataAPI.MktIdxdEvalGet(secID=u"",ticker=ticker,beginDate=u"",endDate=u"",field=u"",pandas="1")
# 市盈率类型：1：TTM市盈率；2：动态市盈率。本例选1
df2=df2[df2.PEType==1]
df2=df2[["tradeDate","PEValue","PB"]]
# 合并行情和因子数据
df3=pd.merge(df1,df2,how='left',on='tradeDate')
df3
```

```python
# 显示PE历史最小值
df3.sort_values(by='PEValue').head()
```

```python
# 转换交易日期为日期类型
df3.tradeDate=pd.to_datetime(df3.tradeDate)
# tradeDate列设为索引
df3=df3.set_index('tradeDate')
# 为显示比例协调，将PB放大500倍，PE放大100倍
df3.PB=df3.PB*500
df3.PEValue=df3.PEValue*100
df3.plot()
```
