# 分级行情分析

```python
import pandas as pd 
import numpy as np

# *************001 返回所有B基金 代码和名称*****************************************************************
bfund = DataAPI.FundLeverageInfoGet(secID=u"",ticker=u"",exchangeCDLeverage=u"XSHG,XSHE",field=u"",pandas="1")

# secIDLeverage,shareType,tickerLeverage,secShortNameLeverage,shareProp,endDate,regSplitDate,upThrshold,downThrshold,navLever,priceLever

bfund = bfund[(bfund.shareType == 'B')&(bfund.endDate.isnull())&(bfund.category == 'E')]

bfund = bfund[['secIDLeverage','tickerLeverage','secShortNameLeverage']]
#bfund

# *************001 返回所有B基金 代码和名称*****************************************************************

myBeginDate0 = '2017-01-17'
myEndDate0 = '2017-03-22'

myBeginDate1 = myBeginDate0.replace('-','')
myEndDate1 =  myEndDate0.replace('-','')


price1 = DataAPI.MktFunddAdjBfGet(secID=bfund.secIDLeverage,ticker=u"",beginDate=myBeginDate1,endDate=myBeginDate1,field=u"secID,ticker,exchangeCD,secShortName,closePrice",pandas="1")

#DataAPI.MktFunddGet(tradeDate=myBeginDate1,secID=bfund.secIDLeverage,ticker=u"",beginDate=u"",endDate=u"",field=u"secID,ticker,exchangeCD,secShortName,closePrice,accumAdjFactor",pandas="1")
#price1 
#secID,ticker,exchangeCD,secShortName,closePrice1

price2 = DataAPI.MktFunddAdjBfGet(secID=bfund.secIDLeverage,ticker=u"",beginDate=myEndDate1,endDate=myEndDate1,field=u"secID,closePrice",pandas="1")

#price2 
#secID,ticker,exchangeCD,secShortName,closePrice1

price3 = pd.merge(price1,price2,how='inner',left_on='secID',right_on='secID')
#price3


#计算变化幅度
price3['CHGPct'] =100*(price3['closePrice_y'] - price3['closePrice_x'])/price3['closePrice_x'] 
price3 = price3.sort(columns=['CHGPct'], ascending=[False])
price3 = price3[['ticker','secShortName','CHGPct']]

price3 = price3[price3['CHGPct']>0]
price3 = price3[price3['CHGPct'] > price3['CHGPct'].mean()]

print '开始日期：'+myBeginDate1+'　结束日期：'+myEndDate1
price3
#price3.head(50)
```
