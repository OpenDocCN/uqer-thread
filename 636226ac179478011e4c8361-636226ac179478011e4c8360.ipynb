# 获取周度数据

```python
zz = DataAPI.MktEquwGet(secID=u'',ticker=u'',weekEndDate=u"",beginDate=u"20211014",endDate=u"20221014",isOpen=u"1",field=u"ticker,secShortName,weekBeginDate,endDate,openPrice,highestPrice,lowestPrice,closePrice",pandas="1")
cc = zz.rename(columns={'ticker':'证券代码','secShortName':'证券名称','openPrice':'周开盘价','highestPrice':'周最高价','lowestPrice':'周最低价','closePrice':'周收盘价'})
print cc 
cc.to_csv('20211014-20221014周度数据.csv')
```
