# 某个日期附近的交易日

```python
def nextDay(currentDate):
    day=int(currentDate[8:10])+1
    month=int(currentDate[5:7])
    year=int(currentDate[0:4])
    if day>=29:
        if day==32:
            day=1
            if month==12:
                month=1
                year=year+1
            else:
                month=month+1
        elif day==31:
            if month in [2,4,6,9,11]:
                day=1
                month=month+1
        elif month==2:
            if (year%4==0 and day==30)or(year%4>0 and day==29):
                day=1
                month=3
    return "%d-%02d-%02d"%(year,month,day)

def tradedate(currentDate="2019-09-18",d=0):
    if d>0:
        while d>0:
            currentDate=nextDay(currentDate)
            dat=DataAPI.TradeCalGet(exchangeCD=u"XSHE",beginDate=currentDate,endDate=currentDate,field=u"",pandas="1")
            if dat.isOpen[0]==1:
                d=d-1
        return currentDate

    dat=DataAPI.TradeCalGet(exchangeCD=u"XSHE",beginDate=currentDate,endDate=currentDate,field=u"",pandas="1")
        
    if dat.isOpen[0]==0:
        currentDate=dat.prevTradeDate[0]    
    if d==0:
        return currentDate
    elif d<0:
        
        while d<0:
            d=d+1
            dat=DataAPI.TradeCalGet(exchangeCD=u"XSHE",beginDate=currentDate,endDate=currentDate,field=u"",pandas="1")
            currentDate=dat.prevTradeDate[0]
            
        return currentDate
    
    return currentDate


print("当日之前最后的交易日")
print(tradedate("2008-05-19"))   
print(tradedate("2008-05-18"))

print("当日往前三个交易日")
print(tradedate("2008-05-19",d=-3))   

print("当日之后第三个交易日")
print(tradedate("1995-01-01",d=3))


```
