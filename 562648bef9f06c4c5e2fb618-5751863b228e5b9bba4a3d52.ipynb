策略思路
---------

* Momentum：业绩好的股票会继续保持其上涨的势头，业绩差的股票会保持其下跌的势头


策略实现
---------

* Momentum：每次调仓将股票按照前一段时间的累计收益率排序并分组，买入历史累计收益 **最高** 的那一组
