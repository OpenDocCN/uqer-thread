**第一部分 有效因子的筛选**
=========================

1.1 对因子策略的理解
-------------------
>一个市场中股票的收益具有一定的分布，这种分布可能对某些参数或者因子有一定的相关性。在可以对市场做空的情况下，寻找到这样的因子，区分出该市场股票中收益较好的一部分并持有，同时做空市场，就可以得到一个稳定的alpha收益。我的研究以HS300成分股为研究样本，对优矿提供的240多个因子（DataAPI.MktStockFactorsOneDayGet函数）的历史数据进行数据分析，其中利用2008年初到2013年底的数据筛选因子、独立性检测以及多因子组合，以2014年以来的数据进行效果检测。

1.2 思路（对于一个因子）
----------------------

>1.为股票动态分组，分组依据为因子的历史值，组号与因子值正相关（或负相关）；
>2.计算各组在测试时期内的收益情况；
>3.计算各组收益与组号的相关性情况；
>4.判断该因子是否与收益有强相关性，是否可产生较高收益，进行筛选。

1.3 具体操作方法（对所有因子同时批量地进行操作）
--------------------------------------------
>1.对2008年初到2013年底的各因子的数据对每股按月取均值，并计算出同一时期每一种因子的股票升序和降序排序；同时计算各因子的NAN值情况，筛选出能保证每个月缺失数据的股票数低于100只的因子，这些因子是我要研究的因子。
>2.按照上一月因子的均值对股票进行分组，本月分别持有这些分组的股票，计算出每组的资金额。由于我是对同时各因子批量处理，考虑到NAN值问题，我仅对排序得到的前200只股票进行分组，共分20组，每10只一组。另外，为防止这种只取前排数据的方法会对正相关因子和负相关因子产生不同影响，我在研究中分别使用了升序(rank)和降序(anti-rank)排序的数据。
>3.计算各因子的每组收益与因子值（用组号代替）的相关性。我分别计算了每年度内的收益相关性情况，并取其平均值和标准差，也计算了全部研究期内的总收益相关性情况。
>4.以‘平均值’、‘标准差’、‘总收益相关系数’和前4组收益均值为筛选参数，筛选出对收益有强相关性的因子，我称之为‘有效因子’。其中：对于降序排序的数据，即‘anti_rank’系列，按照一开始的设计，应该取相关系数为负极大端的因子，这些因子数值与收益成正相关关系，是正相关因子；对于升序排序的数据，即‘rank’系列，应该取相关系数为负极大端的因子，这些因子数值与收益成负相关关系，是负相关因子。
>5.对具有正相关性的有效因子的测试数据进行展示。
>6.以因子‘OperatingExpenseRate’、‘EARNMOM’为例进行简单的回测，证券池为动态的HS300成分股，算法为每日获取上一日各股该因子数值，并持有因子值最大的40只HS300成分。

1.4 研究代码
-----------
>因为该研究代码的运行有点费时，为了防止中途网络中断等问题，每一部分代码都会对计算结果进行储存，下一段代码会重新载入储存的数据。所有数据储存在Multi-factor Model_2文件夹中。以下为每一部分的具体代码。

>1.4.1 对2008年初到2013年底的各因子的数据对每股按月取均值，并计算出同一时期每一种因子的股票升序和降序排序；同时计算各因子的NAN值情况，筛选出能保证每个月缺失数据的股票数低于100只的因子，这些因子是我要研究的因子。

```python
import pandas as pd
from pandas import Series, DataFrame

def read_factor(daytime, stock):  # 读取因子数据的函数，每次只能一个交易日数个股票的数据，注意读取的数据中股票secID码的顺序会重新排序
    f = DataAPI.MktStockFactorsOneDayGet(daytime,secID=stock,ticker=u"",field=u"",pandas="1").set_index(['secID', 'tradeDate']).drop('ticker', axis=1)
    return f

############# 读取交易日数据 ##############
# beginDate=u"20080101",endDate=u"20131231"
date=DataAPI.TradeCalGet(exchangeCD=u"XSHG",beginDate=u"20080101",endDate=u"20131231",field=u"calendarDate,isOpen,isMonthEnd",pandas="1")
date['isMonthStart']=0
number = date[date['isOpen']==1].index
start = True
for i in number:   # 设置月初标记'isMonthStart'
    if start:
        date['isMonthStart'][i] = 1
        start = False
    elif date['isMonthEnd'][i]==1:
        start = True
tradeDate = date[date['isOpen']==1]['calendarDate']
tradeDate.index = tradeDate

############# 各因子历史数据每月均值排序与NAN值统计 #############
mean_factor = DataFrame()
rank_mean_factor = DataFrame()
count_NAN = DataFrame()
count_notNAN = DataFrame()

for day in tradeDate:
    if date[date['calendarDate']==day]['isMonthStart'].any()==1:  # 月初
        stocks = DataAPI.IdxConsGet(secID=u"",ticker=u"000300",intoDate=day,isNew=u"",field=u"",pandas="1")['consID'] # 获取该月（月初交易日）HS300成分股列表
        total_month = read_factor(day, stocks)   # 读取该月首个交易日因子数据
    else:
        factors = read_factor(day, stocks)    # 读取该交易日因子数据
        total_month = pd.concat([total_month, factors], axis=0)   # 将同一月份因子数据沿时间轴合并
        if date[date['calendarDate']==day]['isMonthEnd'].any()==1:  # 月末
            mean_factor_month = total_month.mean(level=0, axis=0, skipna=True)   # 对该月每只股票的各因子数据取平均值，得到本月各股票各因子的平均值，其中NAN值被舍弃掉，不参与均值计算
            mean_factor_month.index = [mean_factor_month.index, [day[:-3]]*300]   # multi-index重设为‘secID’+‘年月’
            rank_mean_factor_month = mean_factor_month.rank(method='first', axis=0)    # 对该月每个因子进行股票排序，排序为升序（ascending=True），结果与factor大小正相关
            mean_factor = pd.concat([mean_factor, mean_factor_month], axis=0)   # 每个月的因子均值沿时间轴合并
            rank_mean_factor = pd.concat([rank_mean_factor, rank_mean_factor_month], axis=0)   # 每个月的因子股票升序排序沿时间轴合并
            print day[:7]   # 输出已完成数据的月份，监视运算进度
# 保存数据
mean_factor.to_csv('Multi-factor Model_2/mean_factor.csv')
rank_mean_factor.to_csv('Multi-factor Model_2/rank_mean_factor.csv')

# 获取各因子每月降序排序的数据（ascending=False），结果与factor大小负相关
all_factor_list = rank_mean_factor.columns.tolist()
anti_rank_mean_factor = DataFrame()
for factor in all_factor_list:
    for_single_factor = rank_mean_factor[factor].unstack().rank(ascending=False, method='first').stack()   #rank结果与factor大小负相关
    for_single_factor.name = factor
    anti_rank_mean_factor = pd.concat([anti_rank_mean_factor, for_single_factor], axis=1)
# 保存数据
anti_rank_mean_factor.to_csv('Multi-factor Model_2/anti_rank_mean_factor.csv')

############## 选择出适合研究的，各月NAN量均小于100股的因子 ################
count_NAN_each_month = rank_mean_factor.isnull().sum(level=1, axis=0)
max_count_NAN = count_NAN_each_month.max(axis=0)
suitable_factors = max_count_NAN[max_count_NAN<100].index.tolist()
# 保存数据
Series(suitable_factors).to_csv('Multi-factor Model_2/suitable_factors.csv')
```

>1.4.2 按照上一月因子的均值对股票进行分组，本月分别持有这些分组的股票，计算出每组的资金额。由于我是对同时各因子批量处理，考虑到NAN值问题，我仅对排序得到的前200只股票进行分组，共分20组，每10只一组。另外，为防止这种只取前排数据的方法会对正相关因子和负相关因子产生不同影响，我在研究中分别使用了升序和降序排序的数据。

```python
import pandas as pd
from pandas import Series, DataFrame

############# 读取交易日数据 ##############
# beginDate=u"20080101",endDate=u"20131231"
date=DataAPI.TradeCalGet(exchangeCD=u"XSHG",beginDate=u"20080101",endDate=u"20131231",field=u"calendarDate,isOpen,isMonthEnd",pandas="1")
date['isMonthStart']=0
number = date[date['isOpen']==1].index
start = True
for i in number:   # 设置月初标记'isMonthStart'
    if start:
        date['isMonthStart'][i] = 1
        start = False
    elif date['isMonthEnd'][i]==1:
        start = True
tradeDate = date[date['isOpen']==1]['calendarDate']
tradeDate.index = tradeDate
month = date[date['isMonthStart']==1]['calendarDate']   # 获取月份格式序列（'year-month'）
month = month.apply(lambda x: x[:-3])
month.index = month

################## 以月为单位对每个因子进行股票分组，计算每组收益 ##########################
factor_list = pd.read_csv('Multi-factor Model_2/suitable_factors.csv', header=None, index_col=0)[1]  # 读取上文中获的适合本研究的因子列表
capital_per_stock = 1  # 每只股票初始分配金额，设置为单1个单位
for rank_type in ['rank', 'anti_rank']:    # ['positive', 'negative'] 对两种排序的数据分别进行测试
    rank_mean_factor = pd.read_csv('Multi-factor Model_2/%s_mean_factor.csv'%rank_type, index_col=['secID', 'Unnamed: 1'])   # 分别读取两种排序的数据
    gathering = [] # 所有factor测验结果汇总
    number = 0
# 开始对每个因子进行测验
    for factor_name in factor_list:
        number = number+1
        print number, factor_name    # 输出当前测试中的因子及编号
        form_1 = Series(10*capital_per_stock, index=range(1, 21))  # 每月月初每组总资金的初始值设定
        form_5 = DataFrame(index=tradeDate.index, columns=range(1, 21)) # 每日每组收盘总资金的初始值设定
        previus_month = '2008-01' # 前一月值初始化，测试中因子应该使用前一个月的数值，防止出现未来数据的干扰
    
        for m in month[1:]:    # 对moth序列进行循环测试
            duration = tradeDate[m+'-00': m+'-32']
            start_day = duration[0]     # 获取每月首个交易日日期形式
            end_day = duration[-1]     # 获取每月最后一个交易日日期形式
        
            form_2 = DataFrame(0, index=range(10), columns=range(1, 21))
            form_2 = (form_2 + form_1/10).T.stack()  # 每月月初每股资金计算
        
            rank_list = rank_mean_factor[factor_name].unstack()[previus_month] # 获取上一月的factor排序
            form_4 = DataFrame(0, index=range(10), columns=range(1,21))  # 每月月初每股前复权开盘价（本月每股持股数），数据初始化
            form_6 = DataFrame(columns=form_4.T.stack().index, index=duration)
        
            for i in range(1, 21):
                rank_by_group = rank_list[(rank_list>(i-1)*10) & (rank_list<=(i*10))].index
                open_close = DataAPI.MktEqudAdjGet(tradeDate=start_day,secID=rank_by_group,ticker=u"",isOpen="",beginDate=u"",endDate=u"",field=['openPrice','closePrice'],pandas="1") #获取月初交易日的openPrice和closePrice
                open_close.ix[open_close['openPrice']==0, 'openPrice']  = open_close[open_close['openPrice']==0]['closePrice'] # 若开盘价为0，替换为对应的closePrice
                form_4[i] = open_close['openPrice']  # 每月月初每股前复权开盘价
                data = DataAPI.MktEqudAdjGet(tradeDate=u"",secID=rank_by_group,ticker=u"",isOpen="",beginDate=start_day,endDate=end_day,field=['tradeDate','secID','closePrice'],pandas="1").set_index(['tradeDate','secID'])['closePrice'].unstack().fillna(method='ffill') # 获取月内每日收盘价，将NAN值以‘ffill’方式替换
                data.columns=range(len(data.columns))
                form_6[i] = data # 本月每日每股收盘价
        
            form_4 = form_4.T.stack() # 每月月初每股前复权开盘价
            form_4 = form_2/form_4 # 本月每股持股数
            form_6 = form_6*form_4 # 本月每日每股资金量
            if form_4.isnull().any(): # 对月初开盘价为NAN值的情况的缺失值处理
                NAN_slice = form_6[form_4[form_4.isnull()].index].copy()
                NAN_slice[:] = 0
                form_6[form_4[form_4.isnull()].index] = NAN_slice + form_2[form_4[form_4.isnull()].index] #将NAN替换为对应的本月初始股票分得资金
            form_5 = form_6.sum(level=0, axis=1, skipna=False) # 每日每组收盘总资金
            if previus_month == '2008-01':
                asset_by_group = form_5
            else:
                asset_by_group = pd.concat([asset_by_group, form_5], axis=0) # 每日每组收盘总资金汇总
        
            previus_month = m # 设置下一月的previus_month值（本月）
            form_1 = asset_by_group.ix[end_day] # 为下一月每组资金赋值

        result_each_factor = asset_by_group
        result_each_factor.columns = [[factor_name]*20, range(1,21)]
        if factor_name == factor_list[0]:  # 将各因子的分组测试数据整合
            gathering = result_each_factor
        else:
            gathering = pd.concat([gathering, result_each_factor], axis=1)

    gathering.to_csv('Multi-factor Model_2/%s_results_suitable_factors.csv' % rank_type)  # 保存数据
```

>1.4.3 计算各因子的每组收益与因子值（用组号代替）的相关性。我分别计算了每年度内的收益相关性情况，并取其平均值和标准差，也计算了全部研究期内的总收益相关性情况。

```python
import pandas as pd
from pandas import Series, DataFrame

############# 读取交易日数据 ##############
# beginDate=u"20080101",endDate=u"20131231"
date=DataAPI.TradeCalGet(exchangeCD=u"XSHG",beginDate=u"20080101",endDate=u"20131231",field=u"calendarDate,isOpen,isMonthEnd",pandas="1")
tradeDate = date[date['isOpen']==1]['calendarDate']
tradeDate.index= tradeDate

endday = '-12-32'
year = range(2008, 2015)   # 该时间范围包含了对2008-2013年每年内的测试数据进行计算，另外最后一次循环中对全部测试数据进行一些统计计算
end_day_year = []
for y in year:
    end_day_year.append(str(y)+endday)

############# 各因子选股收益与因子大小的相关性研究 ###############
type_list = ['rank', 'anti_rank']
for rank_type in type_list:  # 对两种排序方式的数据分别进行研究
    gathering = pd.read_csv('Multi-factor Model_2/%s_results_suitable_factors.csv' % rank_type,header=[0, 1], index_col=0) # 读取各因子的分组测试数据
    gathering.index.name = 'tradeDate'
    gathering.columns.names = ['factor', 'group']
    data_1 = gathering.ix[0]  # 初始化年初各因子各组资金量
    sequence = Series(range(1, 21), index=range(1, 21))  # 构建与因子数值正相关（rank）或负相关（anti_rank）的序列，用于跟分组测试数据进行相关性分析
    correlation_years = DataFrame()  # 初始化相关性计算结果
    for d in end_day_year:  # 以年为单位，研究每一年内的因子分组测试结果与因子的相关性
        if d==end_day_year[-1]:  # 在最后一年的循环中对每年计算的相关系数进行一些统计
            mean = correlation_years.mean(axis=1)  # 几年相关性系数平均值
            mean.name = 'mean'
            std = correlation_years.std(axis=1)   # 几年相关性系数标准差
            std.name = 'std'
            correlation_years = pd.concat([correlation_years, mean, std], axis=1)  # 合并数据
            data_3 = gathering.ix[-1]/gathering.ix[0]   # 计算几年总的相关性系数
            d = 'all'
        else:
            data_2 = gathering.ix[:d].ix[-1]   # 获取每年年末各因子各组资金量数据
            data_3 = data_2/data_1   # 计算每年年末各因子各组资金量相对于其年初值的比值
            data_1 = data_2   # 设置下一年年初资金量（使用本年末资金量）

        p = data_3.unstack().T
        p.index = p.index.astype(int)
        p = p.sort_index(axis=0)  # 规整数据

        correlation = p.corrwith(sequence)   # 计算各因子该年数据与因子大小的相关性
        correlation.name = d[:4]

        correlation_years = pd.concat([correlation_years, correlation], axis=1)
    correlation_years.to_csv('Multi-factor Model_2/%s_correlation.csv' % rank_type)   # 分别保存两种排序的相关性数据
```

>1.4.4 以‘平均值’、‘标准差’、‘总收益相关系数’和前4组收益均值为筛选参数，筛选出对收益有强相关性的因子，我称之为‘有效因子’。其中：对于降序排序的数据，即‘anti_rank’系列，按照一开始的设计，应该取相关系数为负极大端的因子，这些因子数值与收益成正相关关系，是正相关因子；对于升序排序的数据，即‘rank’系列，应该取相关系数为负极大端的因子，这些因子数值与收益成负相关关系，是负相关因子。

>1.4.5 对具有正相关性的有效因子的测试数据进行展示。大图中显示为因子‘OperatingExpenseRate’的测试数据，这里对20组数据进行了相邻4组求均值的处理，蓝、绿、红、灰、紫这几种颜色分别表现了1-4组、5-8组、9-12组、13-16组和17-20组的平均资金量。可以看到各组资金额与组号有着比较明显相关性。下方12个小图为筛选出的12个正相关因子的测试数据。

```python
import pandas as pd
from pandas import Series, DataFrame
import numpy as np
import matplotlib.pyplot as plt
import math

asset_p = pd.read_csv('Multi-factor Model_2/anti_rank_results_suitable_factors.csv',header=[0, 1], index_col=0)
asset_p.index = pd.DatetimeIndex(asset_p.index)
posi_corr_factor_info = pd.read_csv('Multi-factor Model_2/positive_correlated_factor_information.csv', index_col='factor')
factor_list = posi_corr_factor_info.index.tolist()

def n_group_mean(factor, n):
    n_group_mean_results = DataFrame()
    new_group_number = 20/n
    for i in range(new_group_number):
        group_mean = (asset_p[factor][np.arange(i*n+1,i*n+5).astype('str')].mean(axis=1))
        group_mean.name = '%d-%d'%(i*n+1, i*n+4)
        n_group_mean_results = pd.concat([n_group_mean_results, group_mean],axis=1)
    return n_group_mean_results

n_group_mean('OperatingExpenseRate', 4).plot(title='OperatingExpenseRate', figsize=(10,8))

b = 3.0
a = math.ceil(len(factor_list)/b)

fig, ax = plt.subplots(int(a), int(b), sharex=True, sharey=True, figsize=(16,14))
for i in range(int(a)):
    for j in range(int(b)):
        n = i*3+j
        ax[i, j].plot(n_group_mean(factor_list[n], 4))
        ax[i, j].set_title(factor_list[n])
plt.subplots_adjust(wspace=0.1)
```
