相对于回测的时间区间2月~今，本次选取的5月28日股权质押比例样本是“未来数据”，所以作图时只用了两组极端值：质押比例小于10，以及比例大于60。因为我的假设是在二月至今一个多季度的时间跨度内，样本的质押比例不会有本质（大幅）的跳跃

```python
fig, axes = plt.subplots(1, 1, figsize=(8, 5))
sns.distplot(data_10.dailyReturnReinv, bins=25, ax=axes)
#sns.distplot(data_20.dailyReturnReinv, bins=25, ax=axes) 
#sns.distplot(data_40.dailyReturnReinv, bins=25, ax=axes) 
#sns.distplot(data_60.dailyReturnReinv, bins=25, ax=axes) 
sns.distplot(data_80.dailyReturnReinv, bins=25, ax=axes, color='red') 
fig.tight_layout()
```

高股权质押比例->高峰度及左偏（即收益偏负）
蓝线的低峰度可能是因为样本基数大导致（小于10%质押比例的约占总样本数一半）

```python
#区间累积收益率比较
data_10.dailyReturnReinv.cumsum().plot(label='[0,10)')
#data_20.dailyReturnReinv.cumsum().plot(label='[20,40)')
#data_40.dailyReturnReinv.cumsum().plot(label='[40,60)')
data_80.dailyReturnReinv.cumsum().plot(label='[60,)',color='red')
plt.legend()
```

股权质押比例		收益/风险（回撤）比率
[0,10%)			10/15=0.7
[60%,)			 5/15=0.3

结论：两者负相关，质押比例越高，风险越大，收益越小

收尾：
- 正因为股权质押比例相对于回测区间是“未来数据”，且是随时间变化的，故回测跨度不敢放太大，弊端当然是数据可能失真了
- 股权质押也是一种融资杠杆，更完善的分析应该结合债务杠杆一并分析
- 5月底减持新政对股权质押影响如何，是否会因质押比例的不同走出更分化极端的行情，有待观察
