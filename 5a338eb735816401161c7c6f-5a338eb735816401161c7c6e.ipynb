# 1. 三一投资管理公司投资理念
---
三一投资管理公司(Trinity Investment Management Corporation) 创立于 1974 年，原本只提供投资机构客户投资研究建议，自1980 年开始为客户管理投资 组合，1999 年成为美国最大的共同基金及投资管理集团之一的欧本海默基金公司 (Oppenheimer Funds, Inc.)的一员，至2000 年第一季底为止，三一投资管理公司 为客户管理的资产总额达42 亿美元，其投资哲学以价值导向为主，并认为建立成功的价值型投资组合，只需要以单纯而简化的概念及系统性的设计即可。”

# 2. 三一投资管理公司投资理念指标量化
---
三一投资管理公司价值选股策略十分简洁，仅考虑三个价值投资的指标:市盈率、市净率、股利收益率。我们做如下指标量化：
* PE大于0且处于市场最低的30%
* PB处于市场最低的30%
* 股息率处于市场最高的30%
* 前三个条件若选出的股票数量大于30，则选取股息率最高的30只股票
