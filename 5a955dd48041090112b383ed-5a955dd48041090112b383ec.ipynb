进行因子有效性检验ing 所有来自DataAPI.MktStockFactorsOneDayGet数据库的因子都会遇到这个问题，菜鸟痛哭流涕求大神赐教

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
from CAL.PyCAL import *
import functools
from pandas import DataFrame
import scipy.stats as stats
from scipy import stats
from scipy.stats import ttest_ind

df = pd.read_excel('gupiaodaima3.xlsx')
universe =list(df)+set_universe('HS300')
#universe = set_universe('HS300')
#日期列表
date=DataAPI.TradeCalGet(exchangeCD=u"XSHG,XSHE",beginDate=u"20130801",endDate=u"20180101",field=u"calendarDate,isOpen,isMonthEnd",pandas="1")
date=date[date['isOpen']==1]
date=date[date['isMonthEnd']==1]
date=date['calendarDate']
datelist=date.tolist()#结果['2007-08-01', '2007-08-02', '2007-08-03',
datelist=[v.replace('-', '') for v in datelist]
retnmean_main=[]
for dt in datelist:
    retnmean_body=[]
    #i=1
    data=DataAPI.MktStockFactorsOneDayGet(tradeDate=dt,secID=universe,ticker=u"",field=u"secID,tradeDate,EMA5,EMA60",pandas="1")        #取因子
    data=data.dropna()
    
    #EMA5
    EMAF=data.iloc[:,:3]
    EMAF['ticker'] = EMAF['secID'].apply(lambda x: x[0:6])
    EMAF.set_index('ticker',inplace=True)
    #EMA60
    EMAS=data.iloc[:,[0,1,3]]
    EMAS['ticker'] = EMAS['secID'].apply(lambda x: x[0:6])
    EMAS.set_index('ticker',inplace=True)
    EMAF['EMA5']=EMAF['EMA5']/EMAS['EMA60']
    
    data=data.sort(columns='EMA5',ascending=False)
    
    data.index = range(len(data))
    #print data
    data1=data[:int(len(data)*0.2)]
    data2=data[int(len(data)*0.2):int(len(data)*0.4)]
    data3=data[int(len(data)*0.4):int(len(data)*0.6)]
    data4=data[int(len(data)*0.6):int(len(data)*0.8)]
    data5=data[int(len(data)*0.8):]
    retn1=DataAPI.MktEqumAdjGet(secID=data1.secID.tolist(),ticker=u"",monthEndDate=dt,isOpen=u"",beginDate=u"",endDate=u"",field=u"secID,endDate,return",pandas="1")
    retnmean1=retn1['return'].mean()
    retnmean_body.append(retnmean1)
    retn2=DataAPI.MktEqumAdjGet(secID=data2.secID.tolist(),ticker=u"",monthEndDate=dt,isOpen=u"",beginDate=u"",endDate=u"",field=u"secID,endDate,return",pandas="1")
    retnmean2=retn2['return'].mean()
    retnmean_body.append(retnmean2)
    retn3=DataAPI.MktEqumAdjGet(secID=data3.secID.tolist(),ticker=u"",monthEndDate=dt,isOpen=u"",beginDate=u"",endDate=u"",field=u"secID,endDate,return",pandas="1")
    retnmean3=retn3['return'].mean()
    retnmean_body.append(retnmean3)
    retn4=DataAPI.MktEqumAdjGet(secID=data4.secID.tolist(),ticker=u"",monthEndDate=dt,isOpen=u"",beginDate=u"",endDate=u"",field=u"secID,endDate,return",pandas="1")
    retnmean4=retn4['return'].mean()
    retnmean_body.append(retnmean4)
    retn5=DataAPI.MktEqumAdjGet(secID=data5.secID.tolist(),ticker=u"",monthEndDate=dt,isOpen=u"",beginDate=u"",endDate=u"",field=u"secID,endDate,return",pandas="1")
    retnmean5=retn5['return'].mean()
    retnmean_body.append(retnmean5)
    retn0=DataAPI.MktEqumAdjGet(secID=universe,ticker=u"",monthEndDate=dt,isOpen=u"",beginDate=u"",endDate=u"",field=u"secID,endDate,return",pandas="1")
    retnmean0=retn0['return'].mean()
    retnmean_body.append(retnmean0)
    retnmean_main.append(retnmean_body)
new_mean = pd.DataFrame(retnmean_main,index=datelist,columns=['retnmean1','retnmean2','retnmean3','retnmean4','retnmean5','retnmean0'])
    #print dt,retnmean1,retnmean2,retnmean3,retnmean4,retnmean5
#new_mean.at(datelist[0],'retnmean1')#<pandas.core.indexing._AtIndexer at 0x5897e10>
#第二步：计算月平均收益率差——TOP 20%（第一档）- BOTTOM 20%（第五档），TOP 40%（第一、二档）- BOTTOM 40%（第四、五档）
#new_mean.loc['20070831']#new_mean.loc['20070831']得到的是至少两行，从哪到哪
#type(new_mean)#pandas.core.frame.DataFrame
#for i in new_mean.index:
    #retnmeanc1=new_mean.
    #['retnmean1']-new_mean['retnmean5']
#new_mean['retnmeanc1']=[]
#new_mean['retnmeanc1']=
new_mean['retnmeanc1']=(new_mean['retnmean1']-new_mean['retnmean5'])
new_mean['retnmeanc2']=((new_mean['retnmean1']+new_mean['retnmean2'])/2)-((new_mean['retnmean4']+new_mean['retnmean5'])/2)
#第三步：计算上述收益率差的平均值，如果平均值为正，则该因子的影响总体是正向的，即因子值越大收益趋于越大；如果平均值为负，则反之。
new_mean['retnmeancp']=(new_mean['retnmeanc1']+new_mean['retnmeanc2'])/2
#第四步：计算每个因子的有效性：正向(负向)因子的有效性=月均收益率差为正（负）的月份数/总月份数。
positive_num=len(new_mean[new_mean['retnmeancp']>0])
negative_num=len(new_mean[new_mean['retnmeancp']<0])
num=len(new_mean)
ratio=float(positive_num)/num
#print new_mean,ratio #154 48 202
#第五步：根据每个月份五档的因子排名与对应的平均收益率排名，计算其相关系数以及显著性检验的P 值。
mean_all=[]
mean1=new_mean['retnmean1'].mean()
mean_all.append(mean1)
mean2=new_mean['retnmean2'].mean()
mean_all.append(mean2)
mean3=new_mean['retnmean3'].mean()
mean_all.append(mean3)
mean4=new_mean['retnmean4'].mean()
mean_all.append(mean4)
mean5=new_mean['retnmean5'].mean()
mean_all.append(mean5)
new_mean_all=pd.DataFrame(mean_all,index=['mean1','mean2','mean3','mean4','mean5'],columns=['mean_all'])
#print new_mean_all
mean_rank=new_mean_all.rank(axis=0,ascending=False)
mean_rank['nature']=range(1,6)
print mean_rank.corr(),ratio
t,p=ttest_ind(new_mean['retnmean1'],new_mean['retnmean5'])
print t  
print p
new_mean_head=new_mean.drop(['retnmeanc1','retnmeanc2','retnmeancp'],axis=1)
#new_mean_head['HS300']=new_mean['retnmean0'].values
new_mean_head.plot(figsize=(20,10))
```
