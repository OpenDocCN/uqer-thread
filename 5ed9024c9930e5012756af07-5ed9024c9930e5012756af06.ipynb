# Andy_中证500市值与本月涨跌幅相关性

```python
#研究20100101-20191231期间，中证500各股票市值与月度收益率之间的关系。

# 获取月度日历
calendar_frame = DataAPI.TradeCalGet(exchangeCD=u"XSHG,XSHE",beginDate=u"20100101",\
                                     endDate=u"20191231",isOpen=u"1",field=u"calendarDate,isMonthEnd",pandas="1")
calendar_frame = calendar_frame[calendar_frame['isMonthEnd']==1]
calendar_frame.head()
```

```python
#获取中证500该区间的交易数据
tfactor_frame =DataAPI.MktEqudAdjGet(secID=set_universe('ZZ500'),beginDate=u"20100101",endDate=u"20191231",\
                                     field=u"ticker,tradeDate,negMarketValue,closePrice",pandas="1")
tfactor_frame=tfactor_frame[tfactor_frame['tradeDate'].isin(calendar_frame.calendarDate.values)]
tfactor_frame.index = range(len(tfactor_frame))
tfactor_frame.head()
```

```python
#定义向上或者向下拉动函数
def shift_data(df,n):
    df_copy = df.copy()
    df_copy=df_copy.sort_values(by=['tradeDate'],ascending = True)
    df_copy['next_closeprice']=df_copy['closePrice'].shift(n)
    df_copy.dropna()
    return df_copy
```

```python
tfactor_frame_ic = tfactor_frame.groupby(['ticker']).apply(shift_data,-1)
tfactor_frame_ic['return']=(tfactor_frame_ic['next_closeprice']/tfactor_frame_ic['closePrice'])-1
tfactor_frame_ic.index = range(len(tfactor_frame_ic))
tfactor_frame_ic.head()
```

```python
#定义相关函数
import scipy.stats as st
def calc_normal_ic(df):
    tmp_df = df.copy()
    tmp_df.dropna(inplace=True)
    return tmp_df['negMarketValue'].corr(tmp_df['return'])

# 用scipy的包计算
def calc_normal_ic_2(df):
    tmp_df = df.copy()
    tmp_df.dropna(inplace=True)
    return st.pearsonr(tmp_df['negMarketValue'], tmp_df['return'])[0]
```

```python
end_ic = tfactor_frame_ic.groupby(['tradeDate']).apply(calc_normal_ic_2)
end_ic
```
