```python
##两段式分割函数
def ly(x):
    if x>0:
        a = 1
    else:
        a = 0
    return a
```

```python
##三段分割函数
def sy(x):
    if x>1:
        a = 2
    elif x<0:
        a = 1
    else:
        a = 0
    return a
```

```python
data['zd_'] = data['zdf'].apply(ly)
data['zd'] = data['zd_'].shift(-1)
data = data.dropna()
```

```python
choosedata = data['zd']
```

```python
factorlist = list(data.columns)
```

```python
###对所有变量进行划分
dummiesData={}
factordata = choosedata
t = 0
for factor in factorlist:
    try:
        factorData=data.loc[:,factor]
        dummiesData[factor]= pd.qcut(factorData,4,labels = [1,2,3,4])
        factordata = pd.concat([factordata, dummiesData[factor]], axis=1)
        t = t+1
    except:
        continue
```

拿到一个懵逼的不会用的指标，我的第一反应是，如果，这个指标代表一些强弱势，那么，指标区间就可以分段，极强，强势，弱势，极弱。对一个指标分为了四段。所以，我就对所有指标做了四段分段。

```python
###对每个特征打分
scoredata = []
for j in factordata.columns:
    factormean1 = abs(factordata['zd'][factordata[j] == 1].mean()-0.5)
    factormean2 = abs(factordata['zd'][factordata[j] == 2].mean()-0.5)
    factormean3 = abs(factordata['zd'][factordata[j] == 3].mean()-0.5)
    factormean4 = abs(factordata['zd'][factordata[j] == 4].mean()-0.5)
    absmean = factormean1+factormean2+factormean3+factormean4
    scoredata.append(absmean)
```

特征选取第一步就是删除一些无用变量。对变量打分，分数越高变量对其预测越是有用。

```python
###第一步特诊筛选
sfactor = pd.concat([pd.DataFrame(scoredata,columns = ['score']),pd.DataFrame(factordata.columns,columns = ['sfactor'])],axis = 1)
choosefactor = list(sfactor.sort(columns = 'score',ascending = False).sfactor[0:100])
sfactor.sort(columns = 'score',ascending = False)
```

```python
###通过字典方法的变量迭代 onehot编码
choosedict={}
chfactordata = choosedata
t = 0
for factor in choosefactor:
    factorData=data.loc[:,factor]
    choosedict[factor]= pd.get_dummies(pd.qcut(factorData,4,labels = [1,2,3,4]),prefix=factor)
    chfactordata = pd.concat([chfactordata, choosedict[factor]], axis=1)
    t = t+1
chfactordata.index = list(data['tradeDate'])
```

```python
###选取训练集与测试集
chfactordataT = chfactordata[:int(0.75*len(chfactordata))]
chfactordataS = chfactordata[int(0.75*len(chfactordata)):]
```

```python
##显示所取训练集与测试集个数
print len(chfactordataS),len(chfactordataT)
```

```python
from sklearn.svm import LinearSVC
from sklearn.feature_selection import SelectFromModel
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
```

```python
def difflist(s,sub):
    for i in sub:
        if i in s:
            s.remove(i) 
    return s
```

```python
def coef_sort(x,frm):
    data = frm[x]
    index = x[1:]
    ls = LogisticRegression(penalty='l1')
    x = data.iloc[:,1:]
    y = data.iloc[:,0]
    lscf = ls.fit(x,y)
    b1 = pd.DataFrame(lscf.coef_,index = ['coef'],columns= index ).T
    b1 = b1.apply(abs)
    b = b1.sort(columns='coef',ascending= False)
    blist = list(b.index)
    return blist
```

```python
##第二步特征筛选使用逻辑回归
nowscore = 0
trnscore = 0
del_list = []
i = 5
while i < 400:
    xT = chfactordataT.iloc[:,1:i]
    yT = chfactordataT.iloc[:,0]
    xS = chfactordataS.iloc[:,1:i]
    yS = chfactordataS.iloc[:,0]
    ls = LogisticRegression(penalty='l1')
    cllr = ls.fit(xT,yT)
    testscore = cllr.score(xS,yS)
    trainscore = cllr.score(xT,yT)
    if testscore > nowscore or trainscore > trnscore:
        nowscore = max(testscore,nowscore)
        trnscore = max(trainscore,trnscore)
        print i,cllr.score(xT,yT),cllr.score(xS,yS)
    else:
        del_columns = xT.columns[-1]
        del_list.append(del_columns)
    #print(len(x_train.columns))    
    i= i+1
print ('has done')
```

```python
all_column = list(chfactordata.columns[1:])
chooselist_al = ['zd'] + difflist(all_column,del_list)
chfactordata1 = chfactordataT[['zd'] + coef_sort(chooselist_al,chfactordataT[chooselist_al])]
bayes_1 = chfactordataT[chfactordataT['zd'] == 0]
bayes_2 = chfactordataT[chfactordataT['zd'] == 1]
```

```python
##第三步特征筛选通过条件概率选取最优特征数量
i = 5
nowscore = 0
trnscore = 0
del_list = []
while i < len(chfactordata1.columns):
    x = chfactordata1.iloc[:,1:i]
    bayesx_1 = bayes_1.iloc[:,1:i]
    bayesx_2 = bayes_2.iloc[:,1:i]
    #x = x_.drop(del_list,axis = 1)
    y = chfactordata1.iloc[:,0]
    bayesy_1 = bayes_1.iloc[:,0]
    bayesy_2 = bayes_2.iloc[:,0]
    #ls = svm.SVC(kernel='poly',degree = 2)
    #ls = svm.SVC()
    ls = LogisticRegression(penalty='l1')
    #ls = RandomForestClassifier(n_estimators=10)
    cllr = ls.fit(x,y)
    print i,cllr.score(x,y),cllr.score(bayesx_1,bayesy_1),cllr.score(bayesx_2,bayesy_2) 
    i= i+1
print ('has done')
```
