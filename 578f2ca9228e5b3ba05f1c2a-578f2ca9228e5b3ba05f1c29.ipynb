* ## 经济学原理
通常认为股东人数的缩减，意味着持集中度增加则未来票上涨概 认为股东人数的缩减，意味着持集中度增加则未来票上涨概 率大；而股东人数的增加，意味着持更分散票 上涨压力则未来率大；这一结论得出，其实是基于投资者对“主力金”操作手法分析，主力资金在拉升股价之前通常都会进行所谓的“洗盘”和吸筹”，经过盘和吸筹之后，主力以较低的成本获得散户手中码这时候股东人数下降盘和吸筹之后， 主力以较低的成本获得散户手中码这时候股东人数下降持股变得更加集中。随后主力开始拉抬价，在浮盈达到一定水平之持股变得更加集中。随后主力开始拉抬价，在浮盈达到一定水平之始派发筹码，散户投资者成为接盘侠这时候股东人数显著增加持也变得更 始派发筹码，散户投资者成为接盘侠这时候股东人数显著增加持也变得更加分散。

# 二.数据处理和说明
我们要对个股的`股东人数变化`和`户均持股变化`这两个指标进行量化，才能依据其构建投资组合。

根据我们所获得的股东户数以及对应数据的公告日期 ，我们可以整理得到上市公司的月度股东户数和户均持股数的数据 ，每个公司的月度股东数据为`当月月末所能获得的该公司股东户数的最新数据`。

同时，我们知道，通常情况下，对上市公司而言，3月是年报披露日，4月是一季度披露日，8月中报披露，10月三季报披露。那么对于每年而言，1，2月的月度股东户数是一样的，三月变化，四月变化，5，6，7月不变，8月变化，9月不变，10月变化，11，12月都不变。鉴于此，之后我们的投资组合的`调仓周期`也是`每年的3月，4月，8月，10月的月末`(实际建仓是下一个月初第一个交易日)。

也就是说，我们只需要整理出每年3月，4月，8月，10月月末时所能获得的最新的各个公司的股东户和户均持股数，再计算其变化率：
$$股东户数变化率=本期股东户数/上期股东户数-1$$
$$户均持股变化率=本期户均持股数/上期户均持股数-1$$
然后依据计算的结果来构建投资组合，每年调仓四次，观察收益等状况。

```python
#得到每月月末的交易日日期（PS：先取整年的，方便处理，之后再把多余的'20160331'加进去，这里取到2005年，是怕之后改参数，数据不够用）
data = DataAPI.TradeCalGet(exchangeCD=u"XSHG",beginDate='20050101',endDate='20160101',field=['calendarDate','isWeekEnd','isMonthEnd'],pandas="1")
data = data[data['isMonthEnd'] == 1]
date_list = map(lambda x: x[0:4]+x[5:7]+x[8:10], data['calendarDate'].values.tolist())
```

由于我们的调仓周期并不是每月末，所以要对刚才的月末交易日日期进行处理，取出3，4，8，10月月末的交易日日期

```python
#取出3，4，8，10月月末的交易日日期refresh_date
refresh_date=np.zeros(len(date_list)/3)   #一年只有1/3个月会调仓
for i in range(1,len(date_list)+1):      #以index除以12的余数来判断当前index所对应的是几月
    if i%12 == 3:                 #余数为3，表示当前index对应的月是三月，把它存起来，放进refresh_date里，下面一样
        refresh_date[int(i/12)*4+0] = date_list[i-1]
    if i%12 == 4:
        refresh_date[int(i/12)*4+1] = date_list[i-1]
    if i%12 == 8:
        refresh_date[int(i/12)*4+2] = date_list[i-1]
    if i%12 == 10:
        refresh_date[int(i/12)*4+3] = date_list[i-1]
refresh_date=list(refresh_date)        #把refresh_date存为list
refresh_date=[str(int(x)) for x in refresh_date]  
refresh_date=refresh_date+['20160331']   #由于研究时间到了20160401，所以需要把这个数据加进去
refresh_date.sort(reverse=True)       #排序是为了之后会用到，这里是否排序并不影响策略的构建
```

## 2.股东户数和户均持股数数据
接下来我们就要`整理得到个股每一个调仓日时所能获得的最新的股东户数和户均持股数数据`，这里需要用到API（DataAPI.JY.EquHdNumJYGet）

先看看我们的API调出来的数据：

```python
DataAPI.JY.EquHdNumJYGet(ticker='000001',beginDate='20131231',endDate = '20150101',field=u"publishDate,totalHolders,avgHoldShares")
```

totalHolders指的是股东总户数，avgHoldShares指的是户均持股数，publishDate才是我们关注的，表示我们在这个日期才能获得这个数据

正如我们预计的一样，信息每年3，4，8，10月进行披露，但是3月同时会公布一个更近的数据，我们不妨取近的那一个。

先把研究范围内的每只股票的数据存在一个字典里。（存在字典里是因为每只股票的长度不一样，而且本身已经是一个dataframe）

```python
#获取每只个股股东户数和户均持股数的原始数据
stkdata={}  #用于存原始数据的字典，key就是股票的secID,key所对应的是一个dataframe
for i in universe:    #以secID对研究范围内的股票进行循环
    #注意这里输入是ticker
    holdsharesdata=DataAPI.JY.EquHdNumJYGet(ticker=i[0:6],beginDate='20041201',endDate ='20160101',field=u"publishDate,totalHolders,avgHoldShares")
    date=list(holdsharesdata['publishDate'])    #需要对publishDate进行处理，变为年加月份的样子
    date=[str(x) for x in date]
    date=map(lambda x: x[0:4]+x[5:7], date)
    holdsharesdata['publishDate']=date       #把处理好的date重新赋值给publishDate
    holdsharesdata=holdsharesdata.iloc[:,[2,3,4]]    #从上面的例子可以看出，api调用出来的数据还包括ticker和enddate,可以删去
    stkdata[i]=holdsharesdata                 #以secID为key,把该只股票的相关数据赋给它
```

我们不妨看看字典里存的数据是什么样子：

```python
stkdata['000001.XSHE']
```

可以看到，近几年里，3月公布的数据有两个，我们要取近的那个；而早年间的数据，一些publishDate的值是缺失的，我们不妨根据缺失值附近的日期来猜测缺失值是哪个日期。所以我们还要对上面的数据进行处理：

```python
#获取处理后的每只个股股东户数和户均持股数的数据
holddata={}
for i in universe:   #以secID对研究范围内的股票进行循环
    a=stkdata[i]    #调取上面的原始数据
    if int((len(a)-1)/5) == 1:   #如果个股的历史数据有2015年3月的数据，我们就丢掉15年离三月底远那个的数据
        a=a.drop(5)
    elif int((len(a)-1)/5) == 2:  #如果个股的历史数据有2014年3月的数据，我们就丢掉15，16年离三月底远那个的数据
        a=a.drop(5)
        a=a.drop(10)
    elif int((len(a)-1)/5) == 3:  #如果个股的历史数据有2013年3月的数据，我们就丢掉13-16年离三月底远那个的数据
        a=a.drop(5)
        a=a.drop(10)
        a=a.drop(15)
    elif int((len(a)-1)/5) > 3:   #如果个股的历史数据有2012年3月的数据，我们就丢掉12-16年离三月底远那个的数据
        a=a.drop(5)
        a=a.drop(10)
        a=a.drop(15)
        a=a.drop(20)
    try:
        a['publishDate']=refresh_date[:len(a)]    #以之前我们得到的调仓日日期来替换publishDate
        a.set_index(keys='publishDate',inplace=True)  #将调仓日日期设置为index，方便在策略里调用
        holddata[i]=a                      #以secID为key存储股东户数和户均持股数的数据
    except:
        a=a.drop(16)             #这里我发现有几只股票在某一年的2月有三个数据，所以对于这几只股票，我们要多丢一个数据，只取最新的数据
        a['publishDate']=refresh_date[:len(a)]
        a.set_index(keys='publishDate',inplace=True)
        holddata[i]=a
```
