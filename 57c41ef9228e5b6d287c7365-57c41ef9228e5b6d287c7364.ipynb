# 1.考虑噪声的logistic回归原理
***
小弟在论坛上读了一些将机器学习应用在选股上的帖子，主要思路是根据股票收益的大小将其划分为1类和0类，再构建包括基本面和技术面的因子的特征集，训练分类模型，预测股票未来收益是1类还是0类。考虑到股票收益率中的噪声成分很大，将常见的机器学习算法直接应用在选股上可能会有问题(尽管增加正则项可以提高模型的鲁棒性)。本帖参考文献*Label-Noise Robust Logistic Regression and Its Applications* <sup>1</sup>,假设所观测到的标签存在噪声，即观测到的1类标签有一定概率是0类变来的，反之亦然。这样，问题就变成在估计回归系数的同时也估计标签翻转的概率，所要最大化的似然函数就变成
$$  L(\boldsymbol{\omega}) = \prod_{i=1}^{m} p(\tilde{y_i}=1|\boldsymbol{x_i};\boldsymbol{\omega})^{\tilde{y_i}}p(\tilde{y_i}=0|\boldsymbol{x_i};\boldsymbol{\omega})^{1-\tilde{y_i}} \tag{1}$$<br>
$$  p(\tilde{y_i}=k|\boldsymbol{x_i};\boldsymbol{\omega}) = \prod_{j=0}^{1} p(\tilde{y_i}=k|y_i=j)p(y_i=j|\boldsymbol{x_i};\boldsymbol{\omega}) \tag{2}$$<br>
其中，$$p(\tilde{y_i}=0|y_i=1)$$即是1类翻0类的概率。当然，这种噪声只与标签所属类有关的假设非常的理想化，但这种简单的假设可能更安全和稳定。在下文中，我们将这种方法称为Robust Logistic回归。

# 2.一个简单的实验
***
下面是我写的求解Robust Logistic回归的函数，将要优化的目标函数func和其梯度函数func_der，以及初值x0提供给numpy的优化函数fmin_l_bfgs_b，就可以求解了。另外，根据文献1，正则项使用$$({\omega}^2+gamma)^{0.5}$$
来近似ω的绝对值，gamma=1e-8

```python
import numpy as np #数值计算包
import scipy.optimize as opt #优化包
C=1                #正则项系数
gamma=1e-8           
def logit(z):
    '''
        logistic function 即：
        牛顿迭代公式中的p(yi=1 | xi; w)
    '''
    return 1.0 / (1.0 + np.exp(-z)) 
def func(x,XX,YY):
        
        X=XX
        Y=YY
        x=np.mat(x)
        x=x.T
        rows,columns=np.shape(XX)
        w=x[0:columns,0]
        g10=x[columns,0]    #1类flip0类的概率
        g01=x[columns+1,0]   #0类flip1类的概率
        g11=1-g10
        g00=1-g01
        E=np.exp(X*w);                   # 指数乘积
        Item1=np.multiply(Y,np.log(g11*E+g01))   #对数似然函数第一项
       
        Item2=np.multiply(1-Y,np.log(g10*E+g00))   #对数似然函数第二项
        Item3=-np.log(E+1)                 #对数似然函数第三项
        L=-(Item1+Item2+Item3)              #因为调用的优化函数是求最小值，此处取反
        Regulation=C*np.sqrt((w.T*w)+gamma)      #正则项
        # Regulation=C*w.T*w
                  
        L=np.sum(L)+Regulation[0,0]        #加入正则项,得到最终的目标函数
        return L
def func_der(x,XX,YY):
        X=XX
        Y=YY
        x=np.mat(x)
        row,columns=np.shape(XX)
        x=x.T
        w=x[0:columns,0]
        g10=x[columns,0]    #1类flip0类的概率
        g01=x[columns+1,0]   #0类flip1类的概率
        g11=1-g10
        g00=1-g01
        
        E=np.exp(X*w)                          # 指数乘积项
        # 计算回归系数w的梯度
        Item1=np.multiply(Y,np.divide(g11*E,g11*E+g01))    #第一项的梯度
        Item2=np.multiply(1-Y,np.divide(g10*E,g10*E+g00))   #第二项的梯度
        Item3=-E/(E+1)                          #第三项的梯度
       
        L=-X.T*(Item1+Item2+Item3)                  #因为调用的优化函数是求最小值，此处取反
        
        Regulation=C*w/np.sqrt((w.T*w)+gamma)           #正则项的梯度
        
        # Regulation=2*C*rows*w
        
        L=L+Regulation                          #加入正则项
       
        # 计算flip概率的梯度
        grad_g10=np.sum(np.multiply(Y,np.divide(-E,g11*E+g01))+np.multiply(1-Y,np.divide(E,g10*E+g00)))
        grad_g01=np.sum(np.multiply(Y,float(1)/(g11*E+g01))+np.multiply(1-Y,float(-1)/(g10*E+g00)))
        
        L=np.concatenate((L,np.mat(-grad_g10)))
        L=np.concatenate((L,np.mat(-grad_g01)))
        return np.array(L.T)
def rlr(train_X, labels):
 
    X = np.array(train_X) # 训练数据
    rows, columns = np.shape(X)     
    bias=np.ones((rows,1))        
    X=np.concatenate((bias,X),axis=1) #为训练数据增加常数项 
    X=np.mat(X)
    Y = np.mat(labels) # 标签
   
    # 初始回归系数 weights 为全1的向量, weights的行数与输入的训练数据维度相同
    weights = np.ones((columns+1, 1))
    # 初始化1类flip成0类和0类flip成1类的概率
    proba=np.mat([[0.5],[0.5]])
    # 待求解最优问题的初始值
    x0=np.concatenate((weights,proba))
    # 待优化的函数
    
    # 调用fmin_l_bfgs_b求解有约束的最优化问题
    # 设置所求参数的上下界，翻转的概率限定在0~1间，其余无
    mybounds=[]
    for i in range(columns+1):
        mybounds.append((None,None))    # weight 没有上下界
    mybounds.append((0,1))            # flip概率的上下界为1和0
    mybounds.append((0,1))
    
    # 调用fmin_l_bfgs_b求解
   
    x0=np.array(x0.T).reshape(columns+3)
    x,f,d=opt.fmin_l_bfgs_b(func, x0, args=(X,Y), pgtol=1e-3,fprime=func_der,bounds=mybounds,approx_grad=False)
    
    return x
```

在下面的实验中，我们先生成服从二维正态分布，均值(1,1),协方差为单位阵的100个随机数，标记为0类，再生成均值在(1,-1)的100个随机数，标记为1类。通过g10和g01两个变量设置噪声，并用引入噪声后的数据分别来训练Logistic回归模型和Robust Logistic回归模型，最后再生成新的数据对模型的预测能力进行检验。

```python
import numpy as np #数值计算包
import scipy.optimize as opt #优化包
import matplotlib.pyplot as plt #绘图包
import seaborn as sns # 绘图美化包 用于配色
from sklearn.linear_model import LogisticRegression
import pandas as pd


# 生成服从二维正态分布的随机数，0类的均值是[1,1],1类的均值是[1,-1]；协方差矩阵均为单位阵
x0=np.random.normal(1,1,(100,1))
y0=np.random.normal(1,1,(100,1))
X0=np.concatenate((x0,y0),axis=1)
x1=np.random.normal(1,1,(100,1))
y1=np.random.normal(-1,1,(100,1))
X1=np.concatenate((x1,y1),axis=1)

# 0类和1类的标签
Label0=np.zeros((100,1))
Label1=np.ones((100,1))

# 引入噪声，1类flip成0类的概率为g10,0类flip成类的概率是g01
g10=0.5
g01=0.3
g11=1-g10
g00=1-g01
#生成0~1间服从均匀分布的随机变量，用来随机选择翻转的标签，超过阈值g00也即0类维持在0类的概率的随机变量对应的标签将会翻转（好吧，很拗口）
r1=np.random.random(100)
r2=np.random.random(100)
Label0[r1>g00]=1-Label0[r1>g00]
Label1[r2>g11]=1-Label1[r2>g11]

# 将1类和0类的数据组合
X=np.mat(np.concatenate((X0,X1)))
Label=np.mat(np.concatenate((Label0,Label1)))
v=np.where(Label==0)
u=np.where(Label==1)
x0=X[v,0]
y0=X[v,1]
x1=X[u,0]
y1=X[u,1]
# 添加噪声后的图示
plt.figure
plt.scatter(x0, y0, marker='^', c=sns.color_palette()[1])
plt.scatter(x1, y1, marker='o', c=sns.color_palette()[2])
plt.show()
# 应用logistic回归
lr = LogisticRegression(C = 0.1) # C为正则化强度的系数，C越小 模型系数容忍值越大
lr.fit(X, Label)
weights=np.mat(lr.coef_)
bias=np.mat(lr.intercept_)
weights=np.concatenate((bias,weights),axis=1).T

# 鲁棒logistic回归
lr2=rlr(X,Label)
rows,columns=np.shape(X)
Rweight=lr2[0:columns+1]
Rweight=np.mat(Rweight).T
g10=lr2[columns+1]
g01=lr2[columns+2]
print("1类翻0类概率： "),g10
print("0类翻1类概率： "),g01

# 再次生成服从二维正态分布的随机数，作为验证集
x0=np.random.normal(1,1,(100,1))
y0=np.random.normal(1,1,(100,1))
X0=np.concatenate((x0,y0),axis=1)
x1=np.random.normal(1,1,(100,1))
y1=np.random.normal(-1,1,(100,1))
X1=np.concatenate((x1,y1),axis=1)
# 0类和1类的标签
Label0=np.zeros((100,1))
Label1=np.ones((100,1))
X=np.concatenate((X0,X1))
Label=np.concatenate((Label0,Label1))

# 预测并计算错误率
# 普通logistic回归
Forecast1=lr.predict(X)
Fo=Forecast1.reshape(200,1)
ERRate1=float(sum(Fo!=Label))/len(Label)
print("错误率："),ERRate1

# 鲁棒logistic回归
X = np.array(X) # 训练数据
rows, columns = np.shape(X)     
bias=np.ones((rows,1))        
X1=np.concatenate((bias,X),axis=1) #为训练数据增加常数项 
X1=np.mat(X1)
Forecast2=1/(np.exp(X1*Rweight)+1)
Forecast2=pd.DataFrame(Forecast2)
Forecast2=Forecast2.applymap(lambda x: 1 if x>0.5 else 0)
ERRate2=float(sum(np.mat(Forecast2)!=Label))/len(Label)
print("错误率："),ERRate2

plt.figure
plt.scatter(x0, y0, marker='^', c=sns.color_palette()[1])
plt.scatter(x1, y1, marker='o', c=sns.color_palette()[2])

# 画出回归直线： w0 + w1*x1 + w2*x2 = 0, 定义x1为横轴坐标，x2为纵轴坐标
x = np.arange(-4.0, 4.0, 0.1)
y= (-weights[0] - weights[1]*x)/weights[2]
y1,=plt.plot(x, y.T, color=sns.color_palette()[0],Label='Logistic Regression')
y = (-Rweight[0] - Rweight[1]*x)/Rweight[2]
y2,=plt.plot(x, y.T, color=sns.color_palette()[3],Label='Robust Logistic Regression')
plt.legend([y1,y2])
plt.xlabel('x1')
plt.ylabel('x2')
plt.show()
```

可以看到理想化的实验下，Robust Logistic 回归预测能力更好一些，一般是这样。估计的翻转概率分别是1和0.74，跟设定的0.5和0.3不同。实际上，在1类和0类样本数差不多情况下，1类翻0类和0类翻1类两个概率之间的差对模型会有影响，如果概率相同，则影响很小。真正的问题在于，当用在选股上时，即便Robust Logistic 回归能减轻噪声影响，但下期的股票收益率仍然存在噪声，因此表现仍然不确定。那本帖就尝试一下，看能否得到些启发。下面回测的代码是在moo1.0大神的帖子上改动的，在此对他表示感谢，有对代码不清楚的地方，请移步" 机器学习&量化投资-从入门到放弃全套笔记-LogisticRegression ":<https://uqer.datayes.com/community/share/5788af62228e5b8a03932cc5>
