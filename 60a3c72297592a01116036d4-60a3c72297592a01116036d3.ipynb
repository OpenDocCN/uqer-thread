# Macd指标的量化投资策略

```python
def timing_basic_MACD(context):
    # 获得投资者的股票账户fantasy_account
    account = context.get_account("fantasy_account")
    # 获取当前除停牌外的所有可供投资股票（univetse）
    current_universe = context.get_universe(exclude_halt=True)
    
    history = context.history(current_universe, 'closePrice',  max_history_window, rtype='array')
    # log.info(history)
    security_position = account.get_positions()
    # 获取股票账户可用于投资的现金额度 账户中现金
    cash = account.cash
    
    buylist = []
    
    for sec in current_universe:
       
        close = history[sec]['closePrice']
        DIF, DEA, MACD = ta.MACD(close, fastperiod=context.fast_period, 
                                slowperiod=context.slow_period, signalperiod=context.signalperiod)
        
        if MACD[-2]<0 and MACD[-1]>0 and sec not in security_position: # dif 上穿 dea, 且无持仓
            buylist.append(sec)
        elif MACD[-2]>0 and MACD[-1]<0 and sec in security_position:
            account.order_to(sec, 0)  #全部卖出
            # 统计卖出股票的账户金额 context.current_price(sec)获取sec 股票当钱价格
            # security_position[sec].amount 获取股票【sec】的数量
            
    
    log.info(buylist)
    # 交易执行部分
    d = min(len(buylist), int(cash)//100000)  # 每只买10W
    for sec in buylist[:d]:
        account.order(sec, 100000/ context.current_price(sec))
```

```python
def timing(context):
    timing_basic_MACD(context)
```
