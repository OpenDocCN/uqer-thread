# 期货公司持仓排名前20——多空

```python
import datetime
import pandas as pd
from pandas import DataFrame
oneday=datetime.timedelta(days=1)
today=datetime.date.today()
yesterday=today-oneday
beginDate='20170728'
endDate=beginDate
beginDate1='20170731'
endDate1=beginDate1
A=[];B=[];C=[];D=[]
ticker=['bu1709','cu1709','al1709','ni1709','zn1709','rb1710','hc1710','ru1709','ag1712','au1712','a1709','c1709','m1709','y1709','p1709','cs1709',
        'pp1709','v1709','l1709','i1709','j1709','jm1709','jd1709','ma709','sr709','cf709','fg709','ta709','oi709','rm709','zc709',
        'if1708','ic1708','ih1708','tf1709','t1709']
for i in ticker:
    DL=DataAPI.MktFutMLRGet(secID=u"",ticker=i,beginDate=beginDate,endDate=endDate,field=u"",pandas="1")
    DL=DL.sort(['rank'])
    DL=DL.iloc[:20]
    x=DL['longVol']
    jj=x.sum()
    A.append(jj)
Long_sum=pd.DataFrame(A,index=ticker,columns=['Long_sum']) 
#####################################################################################################################################################
for i in ticker:
    KL=DataAPI.MktFutMSRGet(secID=u"",ticker=i,beginDate=beginDate,endDate=endDate,field=u"",pandas="1")
    KL=KL.sort(['rank'])
    a=KL.iloc[:20]
    x=a['shortVol']
    II=x.sum()
    B.append(II)
Short_sum=pd.DataFrame(B,index=ticker,columns=['Short_sum'])
#########################################################################################################################################################
for i in ticker:
    DL_1=DataAPI.MktFutMLRGet(secID=u"",ticker=i,beginDate=beginDate1,endDate=endDate1,field=u"",pandas="1")
    DL_1=DL_1.sort(['rank'])
    a=DL_1.iloc[:20]
    x=a['longVol']
    kk=x.sum()
    C.append(kk)
Long_sum_1=pd.DataFrame(C,index=ticker,columns=['Long_sum_1']) 
#####################################################################################################################################################
for i in ticker:
    KL_1=DataAPI.MktFutMSRGet(secID=u"",ticker=i,beginDate=beginDate1,endDate=endDate1,field=u"",pandas="1")
    KL_1=KL_1.sort(['rank'])
    KL_1=KL_1.iloc[:20]
    KL_1=KL_1['shortVol']
    II=KL_1.sum()
    D.append(II)
Short_sum_1=pd.DataFrame(D,index=ticker,columns=['Short_sum_1'])
DD_sum=pd.concat([Long_sum,Long_sum_1],axis=1)
KD_sum=pd.concat([Short_sum,Short_sum_1],axis=1)
# DD_sum.to_csv('DD_sum.csv')
# KD_sum.to_csv('KD_sum.csv')
DD=DD_sum['Long_sum_1']-DD_sum['Long_sum']
KD=KD_sum['Short_sum_1']-KD_sum['Short_sum']
w=pd.concat([DD,KD],axis=1)
w.columns=['DD','KD']
#w.columns=['多头差','空头差']
w.to_csv('DD_KD_SUM.csv')
#w
```
