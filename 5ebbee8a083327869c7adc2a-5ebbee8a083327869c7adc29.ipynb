```python
#跨期套利
def select_contracts1(date="2019-12-01",kind="CU",sep=[2,3]):
    '''
    月初 选择跨期套利下一个月要进行交易的合约代码
    params:
        date:当前日期
        kind:标的 [CU,ZN,AL,"RB","RU"]
        sep:两个合约交割月与现在的所差月份数
    return:
        两个合约的代码
    '''
    kind=kind.upper()
    d = pd.to_datetime(date)
    d1 = d + pd.Timedelta(15+sep[0]*30,unit="D")
    d2 = d + pd.Timedelta(15+sep[1]*30,unit="D")
    secid1 = kind+d1.strftime("%y%m")#+"."+kind_exchange_dict[kind]
    secid2 = kind+d2.strftime("%y%m")#+"."+kind_exchange_dict[kind]
    return secid1,secid2

def all_contracts1(start="2019-09-01",end="2019-12-01",kind="CU",sep=[2,3]):
    date_list = list(pd.date_range(start=start,end=end,freq="MS"))
    all_contrasts = []
    for date in date_list:
        all_contrasts += list(select_contracts1(date,kind,sep))
    all_contrasts = list(set(all_contrasts))
    return all_contrasts

#test
print select_contracts1(date="2019-12-01",kind="CU",sep=[2,3])
print all_contracts1(start="2019-09-01",end="2019-12-01",kind="CU",sep=[2,3])
```

# 跨期套利

```python
def analysis(bt,perf):
    #根据回测结果，计算相关其他的指标
    trade_seq = bt.futures_trades.sum()
    open_times = len([i for i in trade_seq if trade_seq if i.offset_flag=="open" and i.direction==1])   #开仓次数
    result = dict(
        anual_return = perf['annualized_return'],
        max_drawdown = perf['max_drawdown'],
        volatility = perf["volatility"],
        open_times = open_times,
        r = perf["returns"],
        cumulative_r = perf["cumulative_returns"]
    )
    print(u"anual_return",result["anual_return"])
    print(u"volatility",result["volatility"])
    print(u"max_drawdown",result["max_drawdown"])
    print(u"open_times",result["open_times"])
    
    result["cumulative_r"].plot(figsize=(14,6),fontsize=15)
    return result
```

## 锌

```python
bt_zn = deepcopy(bt)
perf_zn = deepcopy(perf)
result_zn = analysis(bt_zn,perf_zn)
```
