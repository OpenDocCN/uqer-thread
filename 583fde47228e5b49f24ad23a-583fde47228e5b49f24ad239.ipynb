# 高维马格维次优化

# 一.前言
- ## 本帖主要尝试了一种在高维情况下，马格维次均值方差模型的运用，重在实践，理论描述就不写太多了。
- ## 如我们所知，在大类资产的配置上，马格维次均值方差模型效果是不错的，尤其是动态配仓以后，几乎能保持风险很低的稳定的收益。
- ## 然而，在股票市场运用中，我们会发现由马格维次模型所得到的权重，要么是分布太散，要么太集中，总之，结果并非尽如人意，也由此展开了许多研究
- ## 先简单回顾一下马格维次均值方差模型：
$$ max \quad \omega^T\mu $$
$$s.t. \omega^T\Sigma\omega \leq\sigma^2$$
- ## 通过拉格朗日乘子法，我们可以得到上述问题的解，$\omega^* = \frac{\sigma}{\sqrt{\mu^T\Sigma^{-1}\mu}}\Sigma^{-1}\mu$
- ## 可以看到，问题要有解，$\Sigma$就要是非奇异的，但在股票市场里，我们股票的数目是很多的，也不可能取太久之前的数据来计算协方差矩阵，解决这个问题的方法可以有两种，一种就是对协方差阵进行shrinkage估计，一种可以使用Bootstrap的方法.在实际使用的过程中，也要注意权重不能为负值，和为1等问题，这时候就没有一个解析式，常常从局部最优解去找全局最优解
- ## 本帖根据MengMeng AO(2014-2015)等人的论文，复现了一种处理高维情况下马格维次模型的方法，仅供参考

# 二.理论
- ## 关于该方法的理论部分，请阅读论文，这里主要简单介绍我自己的理解并实现的过程
- ## 步骤：
> ## 1.将马格维次均值方差模型转换为回归问题$r^* = \omega^TR+\epsilon$,并且均方误用lasso型的最小平方和形式来处理：
$$min \quad||r_c-X\omega||^2_2$$
$$s.t.\sum|\omega_i|\leq K_p$$
其中，X则为样本矩阵，rc需要用估计量$\hat r_c$
> ## 2.设置不同的$K_P$,使得
$$min \quad |\hat \omega^T\hat \Sigma\omega-\sigma^2|$$,其中$\sigma$是预先设置的风险值

- ## 这样，我们就得到了最优权重

# 三.实践

```python
#加载需要使用的包
from __future__ import division
import numpy as np
import pandas as pd
import scipy.stats as st

from CAL.PyCAL import *    # CAL.PyCAL中包含font
font.set_size(22)
small_font = font.copy()
small_font.set_size(12)
figsize = (14,14*0.518)

import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('white')

from math import *
from sklearn import linear_model
cal = Calendar('China.SSE')
```

```python
class High_dim_Markowitz(object):
    
    def __init__(self,risk,opt_param,begindate,middledate,enddate,universe_name='000300.ZICN'):
        ###----传入参数----
        self.train_begin = begindate  #训练开始时间
        self.train_end = cal.adjustDate(middledate).strftime('%Y%m%d') #训练结束时间
        self.test_bengin = self.train_end      #结束时间
        self.test_end = enddate
        self.risk = risk
        self.opt_param = opt_param
        self.universe = set_universe(universe_name,self.train_end)  #股票池
        
        ###----自定义参数----
        self.weight = self.get_weight()
        self.accumReturns = [0]
        self.indexs = [0]
        
    #计算资产权重
    def get_weight(self):
        return_mat=DataAPI.MktEqudAdjGet(secID=self.universe,beginDate=self.train_begin,endDate=self.train_end, field=u"secID,tradeDate,closePrice")
        X = return_mat.pivot(index='tradeDate',columns='secID', values='closePrice').pct_change().fillna(0.0)
        mu = np.array(X.mean())
        n,p = X.shape
        sigma = X.cov()*250
        tn = 1/(1-p/n)
        kn = p/n/(1-p/n)
        if np.linalg.det(sigma) <0.01 and np.linalg.det(sigma) >-0.01:
            sigma = self.Ledoit_Wolf_Covariance(X)
            invmat = np.mat(sigma).I
        else:
            invmat = np.mat(sigma).I
        theta = np.array(abs(mu.reshape(1,p)*invmat*mu.reshape(p,1)-kn)/tn)[0][0]
        rc = self.risk*(1+theta)/theta
        Y = np.array([rc]*n)
    
        min_standard = 10000000
        best_lam = 0
        best_w = np.array([1/p]*p)
        for i in range(self.opt_param):
            lam = 1/self.opt_param*(i+1)
            clf = linear_model.Lasso(alpha=lam,fit_intercept=False)
            clf.fit(X,Y)
            w = clf.coef_
            standard = np.array(abs(w.reshape(1,p)*np.mat(sigma)*w.reshape(p,1)-self.risk**2))[0][0]
            if standard < min_standard:
                best_lam = lam
                min_standard = standard
                best_w = w
        #调整权重
        best_w = np.array(map( lambda x:max(x,0) ,best_w))
        best_w = best_w/best_w.sum()
        return best_w
    
    #压缩估计的协方差矩阵
    def Ledoit_Wolf_Covariance(self,X):
        n,p = X.shape
        sigma = X.cov()*250
        x = np.mat(X)
        F = np.trace(sigma)/p*np.identity(p)
        up = 0
        for i in range(n):
            up += np.linalg.norm(x[i].reshape(p,1)*x[i].reshape(1,p)-sigma,'fro')**2
        down = n**2*(np.trace(sigma*sigma)-np.trace(sigma)**2/p)
        shrinkage = min(up/down,1)
        return (1-shrinkage)*sigma+shrinkage*F
    
    #从日收益率序列得到累积收益率序列
    def get_accumReturns(self,daily):
        accumReturns=np.zeros(len(daily))
        for i in range(len(daily)):
            tmp = 1.0
            for j in range(i+1):
                tmp *= (1+daily[j])
            accumReturns[i] = tmp-1.0
        return accumReturns
    
    #得到年化收益率
    def get_annualReturns(self,accumReturns):
        annualReturns = (1+accumReturns[-1])**(250.0/len(accumReturns))-1
        return annualReturns
    
    #得到年波动率
    def get_volatility(self,daily):
        volatility = daily.std()*sqrt(250)
        return volatility
    
    #得到夏普比率，rf取3.5%
    def get_sharpe(self,annualReturns,volatility):
        rf=DataAPI.ChinaDataInterestRateInterbankRepoGet(indicID=u"M120000068",beginDate=self.test_bengin,\
                                                                         endDate=self.test_end,field=u"dataValue")
        rf = rf['dataValue'].values.tolist()
        rf = (rf[0]+rf[-1])/2/100/250         
        sharpe = (annualReturns-rf)/volatility
        return sharpe
    
    #得到最大回撤
    def get_maxDrawdown(self,accumReturns):
        drawdown = np.zeros(len(accumReturns))
        for i in range(1,len(drawdown)):
            for j in range(i):
                tmp = np.zeros(i)
                tmp[j] = 1.0-(1+accumReturns[j])/(1+accumReturns[i])
            drawdown[i] = max(tmp)
        maxDrawdown = max(drawdown)
        return maxDrawdown
    
    #得到Calmar比率
    def get_calmarRatio(self,annualReturns,maxDrawdown):
        calmarRatio = annualReturns/maxDrawdown
        return calmarRatio
    
    #定义一个计算投资组合风险指标的函数
    def get_result(self):
        return_mat=DataAPI.MktEqudAdjGet(secID=self.universe,beginDate=self.test_bengin, endDate=self.test_end, field=u"secID,tradeDate,closePrice")
        rtn_test = return_mat.pivot(index='tradeDate',columns='secID', values='closePrice').pct_change().fillna(0.0)
        
        Returns = np.mat(rtn_test.values)*np.mat(self.weight).T               #等权组合的收益率
        Returns = np.array(Returns).reshape(len(Returns),)       #把等权组合的收益率序列处理成数组

        #计算风险指标
        result2 = pd.DataFrame()
        accumReturns = self.get_accumReturns(Returns)    #调用函数计算各项指标
        annualReturns = self.get_annualReturns(accumReturns)
        volatility = self.get_volatility(Returns)
        sharpe = self.get_sharpe(annualReturns,volatility)
        maxDrawdown = self.get_maxDrawdown(accumReturns)
        calmarRatio = self.get_calmarRatio(annualReturns,maxDrawdown)
        result2['Markowitz组合'] = [accumReturns[-1],annualReturns,volatility,sharpe,maxDrawdown,calmarRatio]  #把结果存进dataframe里
        result2.index = ['累计收益','年化收益率','波动率','夏普比率','最大回撤','Calmar比率']
        self.indexs = rtn_test.index
        self.accumReturns = accumReturns
        return result2
    
    #画图
    def get_accumCurve(self):
        accumCurve = pd.DataFrame()
        accumCurve['Markowitz组合'] = self.accumReturns
        accumCurve.index = self.indexs
        ax = accumCurve.plot(figsize=figsize, title=u'累计收益曲线',linewidth=2.5)
        ax.legend([u'Markowitz组合'], prop=small_font)
        ax.title.set_font_properties(font)
        ax.set_axis_bgcolor('white')
        ax.grid(color='gray', alpha=0.2, axis='y')
        return ax
```

- ## risk为期望的风险值，opt_param是处理优化问题时的参数，越大越好，middledate用于划分测试集和训练集

```python
test = High_dim_Markowitz(risk=0.3,opt_param=100000,begindate='20120101',middledate='20140101',enddate='20150101',universe_name='000300.ZICN')
```

```python
test.get_result()
```

- ## 如果把权重打印出来，会发现只有两只股票，其余股票权重为0，这也是为什么要选取lasso型优化，当然，这也是这种方法的弊端，风险会集中在较少的资产上
