# 部分股票无法购买的疑问

```python
#########
#选股函数
#########
import pandas as pd
import numpy as np
from CAL.PyCAL import  *
cal=Calendar('China.SSE')

#porfolo_kind范围为1-5,1表示Q1组合，为收益在前20%的组合
#port_num表示划分的组合总数，此处默认为5个组合
#此策略中，假定排序靠前的收益率更高
#默认为测试PE
#默认股票池为HS300以及ZZ500
def get_buylist(portfoli_kind=1,port_num=5,factorname='PE',conf1=1,index=set_universe('HS300')+set_universe('ZZ500'),currday='2016-09-02'):
    factor_data=DataAPI.MktStockFactorsOneDayGet(tradeDate=currday,secID=index
                                ,field=['tradeDate','secID',factorname]
                                ,pandas='1').set_index('secID').dropna()
    factor_data[factorname]=standardize(neutralize(winsorize(factor_data[factorname])
                                                 ,currday,'short','SW2')).dropna()
    if(conf1):
        factor_data=factor_data.sort(columns = factorname,ascending=True).dropna()
    else:
        factor_data=factor_data.sort(columns = factorname,ascending=False).dropna()        
    stock_num=len(factor_data)
    buylist=list(factor_data.index)[int(stock_num/port_num*(portfoli_kind-1)):int(stock_num/port_num*(portfoli_kind))]
    #buylist = factor_data.tail(int(stock_num/port_num)).index
    return buylist
```
