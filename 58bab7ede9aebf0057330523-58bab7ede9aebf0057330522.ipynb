# 分组分位

对于金融序列的分析经常会用到分组分位的方法，以下是使用这一方法的三部曲：1）等分/分位；2）分组；3）apply函数；

```python
df = DataAPI.MktStockFactorsDateRangeGet(secID=u"",ticker=u"000001",beginDate=u"20150101",endDate=u"20150701",field=u"tradeDate,pe",pandas="1")
df.set_index('tradeDate',inplace=True)
print df.head()#常用的金融序列
print "============"

C = pd.cut(df.PE,10)#第一步
print C.head()
print "============"
G = df.PE.groupby(C)#第二步
print G
print "============"
mean = lambda x:x.mean()
G.apply(mean)#第三步
```

用dataframe的标准金融序列格式可以很方便地使用10等分的数据处理，如上所示，pandas的cut方法等分listlike，后面跟几等分的参数，最后计算每一个等分的平均值；同样的等分使用qcut，如下所示

```python
C = pd.qcut(df.PE,10)#第一步
print C.head()
print "============"
G = df.PE.groupby(C)#第二步
print G
print "============"
mean = lambda x:x.mean()
G.apply(mean)#第三步
```

# 好用的表格

就像Excel里的数据透视表一样，dataframe格式也有些表格函数可以把玩，这里介绍两个，一个pivot_table，另一个是cross_table

```python
data1 = DataAPI.EquIndustryGet(industryVersionCD=u"010303",industry=u"",secID=u"",ticker=u"",intoDate=u"",field=u"secID,outDate,industryID1,industryName1",pandas="1")
data2 = data1.groupby(['secID'])[['outDate']].max() 
data2.reset_index(inplace=True)#重置索引
data3 = DataAPI.MktStockFactorsOneDayGet(tradeDate=u"20170301",secID=u"",ticker=u"",field=u"secID,pe,LFLO,MA20,PB,ROE,",pandas="1")
data4 = pd.merge(data1,data2,on=['secID','outDate'])#并merge在一起
data = pd.merge(data3,data4,on=['secID']).sort('secID')
data.drop(['secID','outDate'],axis=1,inplace=True)
print data.head()
print '==============='
pd.pivot_table(data,columns=['industryID1','industryName1'],values=['PE','LFLO','PB'],aggfunc='median',fill_value=0,margins=True)
```

pivot_table就是一个超级groupby工具，在values下按照columns的值进行groupby，再进行aggfunc函数应用，最后连接起来（很熟悉的套路），里面有几个重要参数：1）index= 可以设置index；2）fill_value=0 如果NA就填0；3）margins=True TRUE的话，每个分组最下面就有个all总体平均值

```python
pd.crosstab(index = data['industryID1'], columns = pd.cut(data.LFLO,3), values=data['LFLO'], rownames=['industryID1'], colnames=['LFLO'], aggfunc='count', margins=True)
```
