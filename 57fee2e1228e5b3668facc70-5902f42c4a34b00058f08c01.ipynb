# 趋同度指标对市场走势的研判

## **趋同度指标对市场走势的研判**
----
### **1.趋同度指标简介**
&ensp;&ensp;&ensp;&ensp;思想来源于曾多次预测海外市场大跌的“兴登堡预兆”信号，据维基百科资料，“兴登堡预兆”的触发条件包括以下四条：
* 当日纽交所（NYSE）中创52周新高以及创52周新低的股票数量同时达到或超过NYSE上市股票数量的2.8%。过往版本使用的阀值为全部交易股票的2.5%。
* NYSE指数点位高于其在50个交易日前的收盘点位。最初的规则为10周移动平均向上，但新规则与上条新高新低规则所用的日数据更为匹配。
* 纽交所创52周新高的股票必须少于创52周新低的股票之两倍；
* 同日McClellan摆动指标为负。

&ensp;&ensp;&ensp;&ensp;其中McClellan摆动指标通过上涨与下跌家数差额定义，为差额的19日EMA（指数移动平均）同39日EMA之差。McClellan以零值为中心，正值对应短期多头市场，负值对应短期空头市场。
<br>
&ensp;&ensp;&ensp;&ensp;我们观察“兴登堡预兆”信号，可以发现其背后的逻辑在于，市场处于极端分化的状态。对于A股市场来说，投机、政策市氛围更浓，故直接使用“兴登堡”信号是不可取的，故我们对“兴登堡预兆”做了本土化。如下所示。
### **2.趋同度指标的构建**
&ensp;&ensp;&ensp;&ensp;为了衡量市场的分化状况，我们使用资本资产定价模型的分析框架。对含有N只股票的市场，每只个股的CAPM均衡模型的表达式为：
$$E(r_i) - r_f = \beta(E(r_m)-r_f),i = 1,2,...,N.$$
&ensp;&ensp;&ensp;&ensp;在CAPM的框架下，个股以及投资组合的风险可划分为系统性风险及非系统性风险两部分。系统性风险与组合$\beta$ 正向相关，反映了市场因子对个股收益造成的影响，即股价波动中趋同的部分。非系统性风险代表组合风险中无法被市场因子解释的部分，即股价波动中分化的部分。
<br>
&ensp;&ensp;&ensp;&ensp;对含有N 个股票的市场，我们定义M 日趋同度指标为:
$$R^2 = \sum_{i=1}^{N}w_iR_i^2$$
&ensp;&ensp;&ensp;&ensp;其中$w_i$代表个股在趋同度指标中所占权重，$R_i$代表每只股票回归的拟合优度。此处，我们取等权重，即平均值。趋同度指标较低时，市场呈现分化特征；趋同度指标较高时，市场呈现趋同的特征。

```python
df_index = DataAPI.MktIdxdGet(ticker=u"000985",beginDate=u"20110803",endDate=u"20170331",field=u"tradeDate,preCloseIndex,closeIndex",pandas="1")
df_index['index_ret'] = df_index['closeIndex'] / df_index['preCloseIndex'] - 1
```

```python
def get_trade_calendar(start_date,end_date):
    """获取交易日历列表"""
    trade_date = DataAPI.TradeCalGet(exchangeCD=u"XSHG",beginDate=start_date,endDate=end_date,field=u"isOpen,calendarDate",pandas="1")
    trade_calendar = trade_date[trade_date['isOpen']==1]
    trade_calendar['calendarDate'] = trade_calendar['calendarDate'].map(lambda x:x.replace('-',''))
    trade_calendar = trade_calendar['calendarDate'].tolist()
    return trade_calendar
trade_calendar = get_trade_calendar('20110803','20170331')
```

```python
def get_component_return(index_symbol,start_date,end_date):
    """获取动态股票池行情"""
    trade_calendar = get_trade_calendar(start_date,end_date)
    df_eq = pd.DataFrame()
    for date in trade_calendar:
        universe = set_universe(index_symbol,date)
        df_tmp = DataAPI.MktEqudAdjAfGet(secID=universe,tradeDate = date,
                                field=u"secID,tradeDate,preClosePrice,closePrice,openPrice",pandas="1")
        df_eq = pd.concat([df_eq,df_tmp],axis=0)
    return df_eq
```

```python
df_eq = get_component_return('000985','20110803','20170331') 
```

```python
df_eq['eq_ret'] = df_eq['closePrice'] / df_eq['preClosePrice'] - 1
df_eq['eq_ret'] = df_eq.apply(lambda x: np.NaN if x['openPrice']==0 or np.isnan(x['openPrice']) else x['eq_ret'],axis=1) #获取个股收益率，停牌则设为缺失值
```

```python
df_eq_ret = df_eq[['secID','tradeDate','eq_ret']]
df_eq_ret = df_eq_ret.set_index(['tradeDate','secID'])
eq_ret = df_eq_ret.unstack()['eq_ret'] # 获取个股历史收益率
```

```python
data = pd.concat([eq_ret,df_index.set_index(['tradeDate'])['index_ret']],axis=1)
data['ii'] = range(len(data))
```

```python
def rsquare(x,y):
    ##计算拟合优度R2
    
    meanofx=x.mean()
    meanofy=y.mean()
    difofxx=x - meanofx
    difofxy=x - meanofy
    difofyy=y - meanofy 

    beta = sum(difofxx * difofyy) / sum(difofxx ** 2)
    r2 = beta**2 * sum(difofxx**2) / sum(difofyy**2)
    return r2
```

```python
def window_convergence(x,data):
    # 计算趋同度
    window_data =  data.iloc[map(int,x)]
    window_data = window_data.dropna(axis=1)
    r2_list = []
    # print list(window_data.columns)
    for col in window_data.columns:
        if col != 'index_ret':
            r2_list.append(rsquare(window_data['index_ret'],window_data[col]))
    #print np.nanmean(r2_list)
    return np.nanmean(r2_list)
```

```python
convergence_30 = pd.rolling_apply(data.ii, 30, lambda x: window_convergence(x, data))
convergence_30.to_csv('convergence_30.csv')
```

#### **3.1 趋同度与历史涨停幅关系**

&ensp;&ensp;&ensp;&ensp;从表中我们可以清晰地看到趋同度和大盘收益率之间存在明显的反向关系，对两者进行回归分析可以发现两者之间的回归直线具有负的斜率值。也就是说市场更多的时候处于分化上涨或趋同下跌的状态，趋同上涨和分化下跌都是不可持续的，会很快存在着变盘。
* 第I 象限：趋同度处于高位，同期大盘上涨，盘中个股趋同性较高。此时的状态是股票普涨，通常不会持续太长，有可能转化到趋同下跌或者分化上涨,但不大可能转化为分化下跌，因为前期有已涨幅，分化意味着还是有涨有跌，总体上涨可能性还很大。
* 第II 象限：盘中个股呈现分化态势，趋同度处于低位，同期大盘指数往上发展。这种情况在熊市时候往往风险已经很大，在牛市就很难说了，有可能转化为趋同上涨。
* 第III象限: 趋同度位于低位，分化程度较大，同期大盘下跌。这种情况在图形中也可以看出是很少的，故很快存在变盘。
* 第IV象限：市场中股票高度趋同，趋同度指数位于高位，同期大盘往往跌幅较大。这种情况在大幅下跌，股灾中很常见，往往会持续一段时间，引起指数的进一步下跌。

#### **3.2 动态分析：周期的循环**
&ensp;&ensp;&ensp;&ensp;上节我们得出的结论是分化上涨、趋同下跌可持续；趋同上涨、趋同下跌不可持续。但是上述结论是基于静态的分析，当我们分析市场时，当时的情况不一定非常明显地处于这四种状态中的其中一种，所以我们还必须了解，当市场从一种状态转化到另外一种状态时中间那段时间市场是如何轮动的。
<br>
&ensp;&ensp;&ensp;&ensp;一个典型的趋同度变化路径2015年9月1日到2016年1月28日为例。
* 缓慢分化：从2015年9月1日到12月16日，趋同度缓慢下降，同期大盘指数逐步加速上升。
* 极端分化：从12月17日到2016年1月4日，趋同度到达底部；指数在同期见顶，开始构造头部。
* 快速趋同：从1月5日到1月9日，趋同度开始快速上涨，盘中个股快速趋同，同期大盘急剧下跌。
* 高度趋同：从1月10日到1月18日，趋同度一路上行，盘中个股呈现高度趋同态势，大盘延续前期跌势，一路下行。

```python
def differentiation_cycle_plot(convergence_data,df_index,start_date,end_date):
    
    start_date = start_date if isinstance(start_date,datetime.datetime) else parse(start_date)
    end_date = end_date if isinstance(end_date,datetime.datetime) else parse(end_date)
    convergence_data_sample = convergence_data[(convergence_data['tradeDate']>=start_date) & (convergence_data['tradeDate']<=end_date)]
    df_index_sample = df_index[(df_index['tradeDate']>=start_date) & (df_index['tradeDate']<=end_date)]
    fig = plt.figure(figsize=(18,7))
    ax = fig.add_subplot(111)
    ax2=ax.twinx()
    ax.plot(df_index_sample['tradeDate'].values,df_index_sample['closeIndex'].values,'darkgoldenrod',linewidth=2.5)
    ax2.plot(convergence_data_sample['tradeDate'].values,convergence_data_sample['convergence_30'].values,'crimson',linewidth=2.5)
    ax.legend([u"中证全指"],loc='upper left',bbox_to_anchor=(0.01,0.99),prop=font,frameon= False,handlelength = 3,handletextpad=0.1,borderpad = 0.1)
    ax2.legend([u"中证全指趋同度"],loc='upper left',bbox_to_anchor=(0.01,0.92),prop=font,frameon= False,handlelength = 3,handletextpad=0.1,borderpad = 0.1)
    ax.spines['top'].set_color('none')
    ax2.spines['top'].set_color('none')
    ax.xaxis.set_ticks_position('bottom')
    ax.axvspan(datetime.datetime(2015,9,1),datetime.datetime(2015,12,17),facecolor='pink', alpha=0.2)
    ax.axvspan(datetime.datetime(2015,12,17),datetime.datetime(2015,12,31),facecolor='lemonchiffon', alpha=0.2)
    ax.axvspan(datetime.datetime(2015,12,31),datetime.datetime(2016,1,7),facecolor='crimson', alpha=0.2)
    ax.axvspan(datetime.datetime(2016,1,7),datetime.datetime(2016,1,28),facecolor='dodgerblue', alpha=0.2)
    ax.set_title(u"图3 市场周期分化示例", fontproperties=font, fontsize=18)
    ax.text(datetime.datetime(2015,10,29),5600,u"缓慢分化",fontproperties=font, fontsize=22,color='crimson',alpha=0.5)
    ax.text(datetime.datetime(2015,12,21),5650,u"极端",fontproperties=font, fontsize=22,color='crimson',alpha=0.5)
    ax.text(datetime.datetime(2015,12,21),5540,u"分化",fontproperties=font, fontsize=22,color='crimson',alpha=0.5)
    ax.text(datetime.datetime(2015,12,31),5650,u"快速",fontproperties=font, fontsize=22,color='crimson',alpha=0.5)
    ax.text(datetime.datetime(2015,12,31),5540,u"趋同",fontproperties=font, fontsize=22,color='crimson',alpha=0.5)
    ax.text(datetime.datetime(2016,1,11),5600,u"高度趋同",fontproperties=font, fontsize=22,color='crimson',alpha=0.5)
```

```python
differentiation_cycle_plot(convergence_data,df_index,'20150901','20160128')
```

&ensp;&ensp;&ensp;&ensp;以上看起来像趋同度指标周期变化的一个典型过程。并且挺符合我们的逻辑，市场开始上涨时通常普涨，当上涨到一定阶段后就开始，分化上涨，当分化到一定程度后，就开始急剧下跌，同时趋同度再极度上升。但是不是所有上涨下跌都能符合这个逻辑呢，这里我们一一看过去。

```python
def convergence_index_analysis_plot(convergence_data,df_index,start_date,end_date,text_list=None):
    
    start_date = start_date if isinstance(start_date,datetime.datetime) else parse(start_date)
    end_date = end_date if isinstance(end_date,datetime.datetime) else parse(end_date)
    convergence_data_sample = convergence_data[(convergence_data['tradeDate']>=start_date) & (convergence_data['tradeDate']<=end_date)]
    df_index_sample = df_index[(df_index['tradeDate']>=start_date) & (df_index['tradeDate']<=end_date)]
    fig = plt.figure(figsize=(18,7))
    ax = fig.add_subplot(111)
    ax2=ax.twinx()
    ax.plot(df_index_sample['tradeDate'].values,df_index_sample['closeIndex'].values,'darkgoldenrod',linewidth=2.5)
    ax2.plot(convergence_data_sample['tradeDate'].values,convergence_data_sample['convergence_30'].values,'crimson',linewidth=2.5)
    ax.legend([u"中证全指"],loc='upper left',bbox_to_anchor=(0.01,0.99),prop=font,frameon= False,handlelength = 3,handletextpad=0.1,borderpad = 0.1)
    ax2.legend([u"中证全指趋同度"],loc='upper left',bbox_to_anchor=(0.01,0.92),prop=font,frameon= False,handlelength = 3,handletextpad=0.1,borderpad = 0.1)
    ax.spines['top'].set_color('none')
    ax2.spines['top'].set_color('none')
    ax.xaxis.set_ticks_position('bottom')
    # plt.ylim(df_index_sample['closeIndex'].min(),df_index_sample['closeIndex'].max())
    ax.set_title(u"中证全指趋同度与中证全指", fontproperties=font, fontsize=18)
    if text_list:
        for item in text_list:
            ax.text(item[0],item[1],u"O",fontproperties=font, fontsize=20,color='blue')
```

```python
convergence_index_analysis_plot(convergence_data,df_index,'20110910','20120910',[(datetime.datetime(2012,3,7),2730),(datetime.datetime(2012,5,29),2500)])
```

从2011年9月到2012年9月的情况来看，2012年3月6，7号出现的趋同度低点，精准的预判了2012年1月到3月这波上涨的顶点，发出风险信号。实际上当在指数上涨过程中，出现趋同度的大幅下跌就是一个危险信号，这符合我们一开始的逻辑。但我们也注意到，趋同度指数在一些时段会出现盘整期，涨幅跌幅都很小，以至于无法用其对市场进行研判。产生这个问题，我觉得最大的问题在于趋同度是取了30天的历史数据，那么这个指标肯定是有很强的延后性的，而且其对小时间段的涨跌的预测研判必然也是失效的。
