```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import datetime as dt
import CAL.PyCAL as C_cal # CAL provides classes and functions dealing with date, time, and calendars. 
from dateutil.parser import parse
import itertools as itl

sns.set_style('white')
pd.set_option('display.width', 200)
cal = C_cal.Calendar('China.SSE')
```

```python
begin_Date = '20180801' # 开始日期
end_Date = (dt.datetime.today() - dt.timedelta(days=1)).strftime('%Y%m%d') # 截止日期

"""
输入行业分类通联编码（如，010303表示申万行业分类2014版）或输入一个行业分类标准名称，获取行业分类标准下行业划分。
DataAPI.IndustryGet(industryVersion=u"SW",industryVersionCD=u"",industryLevel=u"1",isNew=u"1",prntIndustryID=u"",field=u"",pandas="1")

industryVersionCD	list	行业分类标准数字编码。010301-证监会行业V2012、010303-申万行业、010308-中证行业,可以是列表,industryVersion、industryVersionCD至少选择一个
"""
index_symbol = DataAPI.IndustryGet(industryVersion=u"",industryVersionCD=u"010303",industryLevel=u"1",isNew=u"1",field=u"",pandas="1")
index_symbol = index_symbol['indexSymbol'].tolist() 
index_symbol=[str(item) + '.ZICN' for item in index_symbol]  # 加上后缀，使得下面的行情API可以调用

symbol_history_list = [] 
for symbol in index_symbol: # 该API每次只能调用一个indexID
    # 获取指数日线行情信息，包含昨收价、开盘价、最高价、最低价、收盘价、成交量、成交金额等字段。
    df_daily_industry_symbol = DataAPI.MktIdxdGet(ticker=symbol[:6],beginDate=begin_Date,endDate=end_Date,field=u"ticker,tradeDate,closeIndex",pandas="1") # 即，ticker的值是 申万的行业分类代码(如：801010),返回此 分类行业的指数；
    symbol_history_list.append(df_daily_industry_symbol)
    
df_daily_industry_symbol = pd.concat(symbol_history_list,axis=0)  # 将获取的行业数据汇总
df_daily_industry_unstack = df_daily_industry_symbol.set_index(['tradeDate','ticker']).unstack()['closeIndex'] # 将所有的行业行情数据转化为一张二维表

columes_names=['农林牧渔','采掘','化工','钢铁','有色金属','建筑材料','建筑装饰','电气设备','机械设备','国防军工','汽车','电子','家用电器','食品饮料','纺织服装','轻工制造','医药生物','公用事业','交通运输','房地产','银行','非银金融','商业贸易','休闲服务','计算机','传媒','通信','综合']
df_daily_industry_unstack.columns=columes_names
df_daily_industry_unstack.head().append(df_daily_industry_unstack.tail())
```

```python
"""
接下来，我们要将日线的行情数据，转化为月线的行情数据，因为第一天的行情，有的行业存在缺失值，故我们将第一天的行情剔除。定义月行情为每月的第一个交易日到下一月的第一个交易日。
"""
df_daily_industry_unstack = df_daily_industry_unstack.iloc[1:]  # 去掉第一期基期
df_daily_industry_unstack['tradeDate'] = df_daily_industry_unstack.index  # 加入tradeDate列，方便做map

df_daily_industry_unstack['tradeDate'] = df_daily_industry_unstack['tradeDate'].map(lambda x:parse(x)) # 将tradeDate列转化为时间格式
df_daily_industry_unstack['year_month'] = df_daily_industry_unstack['tradeDate'].map(lambda x:(x.year,x.month)) # 得到（年，月）用于筛选
df_daily_industry_unstack = df_daily_industry_unstack.groupby(['year_month']).head(1) # 返回每个月的第一个交易日行情
df_daily_industry_unstack = df_daily_industry_unstack.sort_values(['tradeDate']) # 按tradeDate列排序

del df_daily_industry_unstack['tradeDate'] # 删除tradeDate列和year_month列，方便后面调用后缀函数计算行业收益率
del df_daily_industry_unstack['year_month']

df_daily_industry_unstack
```

```python
"""
接下来我们便要求月度收益率及其排名，调用Series的内置函数pct_change()可以很方便计算收益率；排名调用rank()函数按行做排序就就可以了，默认从小到大。

df.pct_change()
DataFrame.pct_change(periods=1, fill_method=‘pad’, limit=None, freq=None, **kwargs)
表示当前元素与先前元素的相差百分比，当然指定periods=n,表示当前元素与先前n 个元素的相差百分比。
计算变化率：（后一个值-前一个值）／前一个值
"""
monthly_industry_return = df_daily_industry_unstack.pct_change(axis=0) # 调用pct_change()计算收益率
monthly_industry_return = monthly_industry_return.dropna(how='all') # 删除全是缺失值的第一行
monthly_industry_return = monthly_industry_return.rank(axis=1) # 按行做排序
"""
rank 表示在这个数在原来的Series中排第几名，有相同的数，取其排名平均（默认）作为值。

obj = pd.Series([7,-5,7,4,2,0,4])
obj.rank()
0    6.5
1    1.0
2    6.5
3    4.5
4    3.0
5    2.0
6    4.5
dtype: float64

在obj中，4和4的排名是第4名和第五名，取平均得4.5。7和7的排名分别是第六名和第七名，则其排名取平均得6.5
"""

monthly_industry_return
```
