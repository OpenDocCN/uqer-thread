## _基本思想_：  


* 在一定的置信水平下，给定风险敞口水平，可以反算风险资产仓位。  


* 如风险资产组合VaR为5%，设置风险敞口水平为1%，则合理的风险资产仓位是1%/5% = 20%.

```python
# tools
from __future__ import division
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from CAL.PyCAL import *

legendFont = font.copy()
titleFont = font.copy()
legendFont.set_size(14)
titleFont.set_size(20)

sns.set_style('white')
```

* 用到的一些ETF是：

```python
DataAPI.FundGet(secID = ['510300.XSHG','510500.XSHG','159915.XSHE','510050.XSHG', '160216.XSHE','518880.XSHG',
                         '511010.XSHG','511220.XSHG', '513100.XSHG','513500.XSHG','159920.XSHG', '511880.XSHG'], 
                field = 'secID,secShortName,listDate,category').set_index('secID').sort('category').T
```

## _回测设置与结果_：


* 将置信水平confidenceLevel设为99%，风险敞口targetRiskLevel设为1%，计算风险资产沪深300ETF (510300.XSHG)的仓位水平。  


* 备选的无风险资产是5年国债ETF和银华交易型货币ETF，如果5年国债ETF最糟糕20%情况下收益高于0，则选择国债ETF，否则持有货币ETF.
  
  
* 2013年8月至今实现12.2%的年化收益，最大回撤7.7%. 看起来是比较稳定的改进固定收益类策略。

## _持仓与交易情况_：

* 为了避免频繁调仓带来的交易或盯市成本，只有在需要卖/买超过5%时下单，这使得策略调仓频率不高，适合日常操作。
  
  
* 可见债券ETF持仓总体维持在较高水平，稳定为主，也相对改进固收类资产的收益；看来实盘中也不妨一试。

```python
'''根据bt和perf画出回测累计收益、风险资产持仓、换手率情况'''
if not bt.index.name == 'tradeDate':
    bt.set_index('tradeDate', inplace = True)
bt['trade_vol'] = bt.blotter.apply(lambda x: sum([y.filled_amount*y.transact_price for y in x]))
bt['buy_vol'] = bt.blotter.apply(lambda x: sum([y.filled_amount*y.transact_price if y.direction==1 else 0 for y in x]))
bt['sell_vol'] = bt.blotter.apply(lambda x: sum([y.filled_amount*y.transact_price if y.direction==-1 else 0 for y in x]))

fig = plt.figure(figsize = (16, 8))    
ax1 = fig.add_subplot(211)

ax12 = bt.portfolio_value.plot(ax = ax1, secondary_y = True)
res = {}
for x,date in enumerate(bt.index):
    res[date] = bt.security_position[x]
(pd.Panel(res).major_xs('value') / bt.portfolio_value).T.plot(ax = ax1)
ax12.legend([u'资产组合净值'], prop = legendFont)
ax1.legend([u'股票ETF仓位', u'国债ETF仓位'], prop = legendFont, loc = 2)

ax2 = fig.add_subplot(212)
tmpDf = (bt[['trade_vol', 'buy_vol', 'sell_vol']].T /bt.portfolio_value).T
tmpDf['number'] = range(len(tmpDf))
tmpDf = tmpDf[(tmpDf.trade_vol > 0) | (tmpDf.count == 0) | (tmpDf.count == len(tmpDf))]
tmpDf.drop('number', axis = 1, inplace = True)
tmpDf.index = map(lambda x: x.strftime('%Y-%m-%d'), tmpDf.index)
barWidth = 0.1
tmpDf.plot(kind = 'bar', ax = ax2)
# fig.autofmt_xdate(rotation = 30, ha = 'right')
plt.xticks(rotation = 30, ha = 'right')
ax1.grid(axis = 'y')
ax2.grid(axis = 'y')
ax2.legend([u'交易换手率',u'买单换手率',u'卖单换手率'], prop = legendFont, loc = 0)

fig.tight_layout()
plt.show()
```

## _结论_：

* 采用VaR仓位控制方法时，如果设置高置信水平与低风险敞口，可以达到类似固定收益类资产的稳定表现，收益也相对有所改观。  



* 在策略中，为符合实盘条件，对于交易量不超过5%的调整不予下单，与逐日调整结果无明显差别；这意味着策略管理成本不高，有可操作性。
