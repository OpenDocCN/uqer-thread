## 新股首次破板状态统计分析
------
首先我们要找到新股首次破板的日期是哪天，这里我们所说的首次破板是新股除第一天后的第二天开始算起，没有一字板的第一天。得益于通联数据提供的tick级数据计算的涨跌停次数API，我们可以很方便判断一只股票是否在当天是一字板，判别标准为当天开盘价等于当天涨停价，且当天涨停次数等于1。

```python
def get_new_equity_windows_df(windows):
    trade_window = str(windows) + 'B'
    security_listDate = DataAPI.EquGet(equTypeCD=u"A",secID=u"",ticker=u"",listStatusCD=u"",field=u"secID,ticker,ListDate",pandas="1")
    new_equity = security_listDate[security_listDate['listDate'] > '2015-01-01'].reset_index(drop='False')
    new_equity = new_equity[~new_equity['secID'].isin(['603393.XSHG','002808.XSHE'])] ## 这两只股票有毒，先放一放
    new_equity['date_list'] = new_equity['listDate'].map(lambda x:cal.bizDatesList(x,cal.advanceDate(x,trade_window)))
    eq_analysis_list = []
    for i in range(len(new_equity)):
        eq_analysis_list.append(pd.DataFrame([[new_equity.iloc[i]['listDate']]*(windows+1),new_equity.iloc[i]['date_list'],[new_equity.iloc[i]['secID']]*(windows+1)]).T)
    new_equity_df = pd.concat(eq_analysis_list,axis=0)
    new_equity_df = new_equity_df.reset_index(drop=True).rename(columns = {0:'listDate',1:'trade_date',2:'secID'})
    new_equity_df['trade_date'] = new_equity_df['trade_date'].map(lambda x:x.strftime('%Y-%m-%d'))
    return new_equity_df
```

```python
new_equity_df = get_new_equity_windows_df(50) #得到新股发行后的50个交易日，没有新股超过50个交易日的一字板涨停
```

```python
new_equity_norm_df = normalize_event(new_equity_df,publish_date_field='trade_date',security_field='secID')
```

```python
new_equity_base_df = data_infer.build(new_equity_norm_df).append('BASE_DATE').get()
new_equity_base_df = data_infer.build(new_equity_base_df).append('UpLimitNum').append('limitUpPrice').append('openPrice').get()
```

```python
def break_point(x):
    if x.iloc[0]['UpLimitNum'] !=1 and not np.isnan(x.iloc[0]['UpLimitNum']): ##上市日首日就没一次性涨停
        return x.iloc[0]
    for i in range(1,len(x)):
        if x.iloc[i]['openPrice']!=x.iloc[i]['limitUpPrice'] or x.iloc[i]['UpLimitNum']!=1: 
            return x.iloc[i]
```

```python
new_equity_break_df = new_equity_base_df.groupby(['_SECURITY_ID']).apply(lambda x:break_point(x))
new_equity_break_df = new_equity_break_df.dropna(how='all') #剔除最新数据之前还是一直一字涨停的股票
```

```python
new_equity_break_df['ii'] = range(len(new_equity_break_df))
new_equity_break_df[u'一字板天数'] = new_equity_break_df['ii'].map(lambda x:cal.bizDatesNumber(new_equity_break_df.iloc[x]['listDate'],new_equity_break_df.iloc[x]['BASE_DATE']))
```

我们把这几只股票剔除，继续分析。

```python
new_equity_break_df_1 = new_equity_break_df[new_equity_break_df[u'一字板天数']>1]
new_equity_break_df_1 = data_infer.build(new_equity_break_df_1).append('industry_l1',industry=u'申万').get() ##上市几日的行业分类存在缺失，我们直接用破板时候的行业分类，影响不大
```

```python
##部分股票上市第一天的因子数据缺失，此处我们用上市第二天的因子数据进行进一步分析
new_equity_break_df_1['ADJUST_DATE'] = new_equity_break_df_1['listDate'].map(lambda x:cal.advanceDate(x,'1B').strftime('%Y%m%d'))
```

```python
#将基准日重新设置为上市第二天
listDate_break_df_1 = normalize_event(new_equity_break_df_1,publish_date_field='ADJUST_DATE',security_field='secID')
listDate_break_df_1 = listDate_break_df_1.drop('BASE_DATE',axis=1)
```

```python
new_equity_break_df_summary = data_infer.build(listDate_break_df_1).append('BASE_DATE').append('LFLO').append('LCAP').append('PE').get()
new_equity_break_df_summary[u'总市值(亿)'] = new_equity_break_df_summary['LCAP'].map(lambda x:math.exp(x)/1e+08)
new_equity_break_df_summary[u'流通市值(亿)'] = new_equity_break_df_summary['LFLO'].map(lambda x:math.exp(x)/1e+08)
```

```python
new_equity_break_df_grouped = event_service.group_by_field(new_equity_break_df_summary,u'流通市值(亿)',[2,5,10,20,50,100])
```

```python
def group_bar_plot(new_equity_break_df_grouped, field):
    f, ax = plt.subplots(1,1,figsize=(10,7))
    neg_mv_summary = new_equity_break_df_grouped.apply(lambda x:x[u'一字板天数'].mean())
    neg_mv_summary.sort_values().plot(kind='bar',ax=ax)
    ax.set_xticklabels([item.replace(field,'') for item in neg_mv_summary.sort_values().index],fontproperties=font)
    ax.set_xlabel(field,fontproperties=font)
    ax.set_title(u'一字板天数',fontproperties=font, fontsize=18)
    plt.setp(plt.gca().get_xticklabels(), rotation=45, horizontalalignment='right')
    plt.show()
```

```python
group_bar_plot(new_equity_break_df_grouped,u'流通市值(亿)')
```

可以看出，不同的流通市值在一字板天数上有显著差异，呈现出明显的小市值效应，需要注意的是此处都是用新股上市第二天的收盘价计算的属性。中了流通市值小于5的新股的就偷着乐了吧。

```python
new_equity_break_df_grouped = event_service.group_by_field(new_equity_break_df_summary,u'总市值(亿)',[5,10,20,50,100])
group_bar_plot(new_equity_break_df_grouped,u'总市值(亿)')
```

用总市值来分组也体现出明显的小市值效应，但可以看出以新股上市第二天的收盘价的总市值，已没有股票是是低于5亿的了。

```python
new_equity_break_df_grouped = event_service.group_by_field(new_equity_break_df_summary,'PE',[20,40,60,100])
group_bar_plot(new_equity_break_df_grouped,'PE')
```

可以看出用PE来进行划分，各股票的一字板数差异不大，可能的原因是各个股票的估值更加参考的各个股票所处的行业，故我们进一步以行业来做分组。

```python
industry_summary = new_equity_break_df_summary.groupby(['申万industry_l1']).apply(lambda x:x[u'一字板天数'].mean())
```

```python
f, ax = plt.subplots(1,1,figsize=(15,7))
industry_summary.sort_values().plot(kind='bar',ax=ax)
ax.set_xticklabels([ item.decode('utf-8') for item in industry_summary.sort_values().index],fontproperties=font)
ax.set_xlabel(u'申万一级行业',fontproperties=font)
ax.set_title(u'一字板天数',fontproperties=font, fontsize=18)
plt.show()
```

从行业属性来看，计算机、国防军工等股票平均来说有更多的一字板，而银行、非银金融、采掘等传统行业则更少的一字板天数。当然，也很有可能是因为银行等行业的股票上市时就市值较大，而计算机等则比较小，内生原因可能还是市值。但也正是因为计算机、通信等属于新兴行业，它们目前公司的都处于起步，所以市值较小；而银行等传统行业，已经非常成熟，上市的公司也市值相对较大。所以也很难说明，这就不是行业的原因。

此处，我们不进一步讨论，行业，市值与估值的交叉影响。我们只要知道，行业与市值对新股上市的一字板天数的影响是非常显著的，而估值从全市场范围来看影响并不是很显著，这对我们进一步对次新股的操作是非常有帮助的。

## 次新股的短期投资机会
-----
此处，次新股的定义是上市首日一次性涨停后，第一次出现不是一字板的股票，这样的股票也是我们实际中可以买得到的股票。而如何把握次新股的短期投资机会呢，有一句叫什么，对一只的股票最好的定价，就是找到一个相似的股票。故短期来说，可能会有某一只股票因整体的市场环境不好而提前破板，这时候参考其相似公司的涨停板数，就可以对该次新股有一个较好的参考意义。当然本身市场环境的变化，就会直接影响新股的涨停天数，但即使至今，新股的涨停逻辑仍在，所以当有某只股票意外破板，就可以把握该机会，果断买入。
<br>
这里不用到未来数据，做一个简单的回测，用2015年6月3日之前破板的99只股票作为初始已知样本，其后破板的350只做为回测股票滚动寻找初始状态最为接近的样本。


```python
new_equity_break_df_summary['break_day'] = new_equity_break_df_summary.index.map(lambda x:cal.advanceDate(parse(new_equity_break_df_summary.ix[x]['listDate']),str(new_equity_break_df_summary.ix[x][u'一字板天数']-1)+'B'))
new_equity_break_df_summary = new_equity_break_df_summary.sort_values(['break_day'])
```

```python
new_equity_break_df_summary['ii'] = range(len(new_equity_break_df_summary))
backtest_summary = new_equity_break_df_summary.iloc[99:]
```

```python
def get_same_industry(x,new_equity_break_df_summary):
    temp_sample = new_equity_break_df_summary.iloc[0:x]
    return temp_sample[temp_sample['申万industry_l1'] == new_equity_break_df_summary.iloc[x]['申万industry_l1']][['listDate','secID',u'一字板天数',u'总市值(亿)',u'流通市值(亿)','PE']]
```

```python
backtest_summary['same_industry'] = backtest_summary['ii'].map(lambda x:get_same_industry(x,new_equity_break_df_summary))
```

```python
backtest_summary.iloc[0:1]
```

```python
backtest_summary['same_industry'][0]
```

这里给出全部信息，实际上对于单一股票，拿出之前相同行业的股票进行综合分析就可以，譬如说上面的例子，`603989`在2015年5月15日上市，在2015年6月3日破板，历经14个涨停板，已经是之前所有电子行业上市的股票的最多涨停天数了，并且流通市值和总市值在之前上市的电子行业中也较大，所以这样一个涨停次数已经算非常多了。也就是在牛市最后的疯狂才有这样一个情况，可以看下K线，股灾迅速跌去了三分之二，这样的股票在破板之后最好不要碰。

这里我们做一个简单的量化，选出破板的股票，其一字板数是之前上市的同行业股票最少的20%的,并且流通市值(亿)、总市值(亿)也是最小50%,PE不要最大的20%。

```python
select_event = []
for i in range(len(backtest_summary)):
    if backtest_summary.iloc[i][u'一字板天数'] < backtest_summary['same_industry'].iloc[i][u'一字板天数'].quantile(0.2) and backtest_summary.iloc[i][u'流通市值(亿)'] < backtest_summary['same_industry'].iloc[i][u'流通市值(亿)'].quantile(0.5) and backtest_summary.iloc[i][u'总市值(亿)'] < backtest_summary['same_industry'].iloc[i][u'总市值(亿)'].quantile(0.5) and backtest_summary.iloc[i]['PE'] < backtest_summary['same_industry'].iloc[i]['PE'].quantile(0.8):
        select_event.append([backtest_summary.iloc[i]['secID'],backtest_summary.iloc[i]['break_day']])
```

```python
select_event_df = pd.DataFrame(select_event).rename(columns={0:'secID',1:'trade_date'})
select_event_df['trade_date'] = select_event_df['trade_date'].map(lambda x:x.strftime('%Y-%m-%d'))
```

```python
select_event_df
```

OK，我们大致选出了这几只股票有比较好的投资机会，有点少，当然你可以放松条件得到更多。现在我们调用事件驱动时间窗口分析，对这几只股票进行事件窗口收益率测试。

```python
select_event_df = normalize_event(select_event_df,publish_date_field='trade_date',security_field='secID')#标准化事件
select_event_df = data_infer.build(select_event_df).append('BASE_DATE').get()
```

```python
select_event_df
```

```python
event_window_gains = event_service.summary_window_gains(select_event_df,range(-1,10),'000300',start_point=0) #最后一只股票最近上市破板，时间窗口不能选太长
event_service.plot_window_gains(event_window_gains)
```

```python
event_window_gains = event_service.summary_window_gains(select_event_df.iloc[0:-1],range(-1,25),'000300',start_point=0) #去除最后一只，我们选取长一点的时间窗口
event_service.plot_window_gains(event_window_gains)
```

可以看出这样选出的次新股在短期确实有比较多的超额收益，观察基准走势可以发现，指数一开始也是上涨的，观察事件列表也可以知道，大部分都是在2015年7月8日的大跌提前破板的，之后指数开始反弹，这些次新股也开始反弹，并且反弹力度超乎想象。
