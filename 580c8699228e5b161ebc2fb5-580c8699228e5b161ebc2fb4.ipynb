### Distribution plots 分布绘图

- jointplot 单因素、双因素 变量图
- pairplot 数据集成对关系
- distplot 单变量分布图
- kdeplot 拟合
- rugplot 轴向柱线，地毯图


```python
import numpy as np
import pandas as pd
from scipy import stats,integrate
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tmp=DataAPI.MktIdxdGet(ticker=u"000001",beginDate=u"2005-01-01",endDate=u"2016-09-30",field=['openIndex','closeIndex'],pandas="1")
tmp1=tmp.diff(-5)/tmp
tmp1=tmp1.dropna()
```

```python
plt.figure(figsize=(26,9))
sns.distplot(tmp['closeIndex'],kde = False, rug= True)
# kde 拟合内核密度
# rugplot() 地毯图
# displot(rug 地毯图)
```

```python
# kw 宽度，参数孔辉估计拟合的紧密程度
# cut 控制对极值追溯程度
sns.kdeplot(tmp['closeIndex'])
sns.kdeplot(tmp['closeIndex'],bw= 0.3 , label= '0.33')
sns.kdeplot(tmp['closeIndex'], bw=3 , label = '3')
```

```python
# 局部设定图形风格
with sns.axes_style('white'):
    plt.figure(figsize=(12,9))
    sns.distplot(tmp1['openIndex'])
```

```python
# 局部设定图形风格
with sns.axes_style('dark'):
    sns.jointplot(x='openIndex',y='closeIndex',data=tmp1)
```

```python
# 六边分箱绘图
plt.Figure(figsize=(16,16))
#with sns.axes_style('dark'):
sns.jointplot(x='openIndex',y='closeIndex',data=tmp1[-200:],kind='hex',color='k')
```

```python
sns.jointplot(x='openIndex',y='closeIndex',data=tmp1[-200:],kind='kde')
```

```python
f,ax = plt.subplots(figsize = (16,16))
sns.kdeplot(tmp1['closeIndex'],tmp1['openIndex'],ax=ax)
```

```python
f,ax = plt.subplots(figsize=(9,9))
cmap = sns.cubehelix_palette(as_cmap=True,dark= 0,light=1,reverse=True)
sns.kdeplot(tmp1.closeIndex[-200:] ,tmp1.openIndex[-200:] ,n_levels=60,shade=True)
```
