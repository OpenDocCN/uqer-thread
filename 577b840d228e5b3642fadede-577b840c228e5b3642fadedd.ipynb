```python
class benchmark():
    def __init__(self,idxID=''):
        if idxID:
            self.idxID = idxID
            self.data = DataAPI.MktIdxdGet(tradeDate=u"",indexID=self.idxID,ticker=u"",beginDate=u"",endDate=u"",field=u"",pandas="1")
        else:
            self.idxID = '000300.ZICN'
            # self.idxID = '000009.ZICN'
            self.data = DataAPI.MktIdxdGet(tradeDate=u"",indexID=self.idxID,ticker=u"",beginDate=u"",endDate=u"",field=u"",pandas="1")
        self.data['tradeDate'] = pd.to_datetime(self.data['tradeDate'],format='%Y-%m-%d')
        
    def compare(self,secID=''):
        lower_dt = (Date.todaysDate() - Period('2Y')).toDateTime()
        fig,ax = plt.subplots(1,1,figsize=(12,6))
        plt.grid()
        if secID:
            df = DataAPI.FundNavGet(dataDate=u"",secID=secID,ticker=u"",beginDate=u"",endDate=u"",field=u"",pandas="1")
            df['publishDate'] = pd.to_datetime(df['publishDate'],format='%Y-%m-%d')
            df = df[df['publishDate'] > lower_dt]
            self.data = self.data[self.data['tradeDate'] > lower_dt]
            ax.plot(df['publishDate'],(1+df['ADJUST_NAV'].pct_change()).cumprod(),label = df['secID'].iloc[0]) # 注意危险，cumnav和adjust_nav的算法和差异来源……
            ax.plot(self.data['tradeDate'],(1+self.data['closeIndex'].pct_change()).cumprod(),label = self.data['indexID'].iloc[0])
            plt.legend()
            plt.show()
        else:
            print("error")
        return 
```

```python
ref = benchmark()
ref.data.set_index('tradeDate',inplace=True)
ref.data = ref.data.resample('m',how='last')
idx = ref.data['closeIndex'].pct_change()
ttlist = []
for secID in efund['secID']:
    if re.search(r'^[0-9]*\.OFCN',secID): # 上面提到的FundGet数据错误的自我修正
        try:
            df = DataAPI.FundNavGet(dataDate=u"",secID=secID,ticker=u"",beginDate=u"",endDate=u"",field=u"",pandas="1")
            if not df.empty:
                secShortName = df['secShortName'].iloc[0]
                df['endDate'] = pd.to_datetime(df['endDate'],format='%Y-%m-%d')
                ser = df.set_index('endDate').resample('m',how='last')['ADJUST_NAV'].pct_change() # 获取基金的月度收益率，注：保证顺序，并且收益率受到复权数据精度的影响。
                ser = ser.dropna() # 本质上是删除第一行。因为pct_change算出来的第一个肯定是NAN
                if not ser.empty:
                    mg = pd.concat([idx,ser],axis=1).dropna()
                    secStd = mg['ADJUST_NAV'].std() # 基金收益率的方差
                    secMeanProfit = mg['ADJUST_NAV'].mean() # 基金的平均收益率
                    baseStd = mg['closeIndex'].std()
                    baseMeanProfit = mg['closeIndex'].mean() # 获取对应时间内，基准指数的平均收益
                    fin = (1 + ser).cumprod() - 1
                    secCumProfit = fin.iloc[-1] # 最终收益倍数，并非百分比
                    ttlist.append([secID,secShortName,secCumProfit,secMeanProfit,secStd,baseMeanProfit,baseStd])
                    
        except Exception as e:
            print(df,secID)
            
df = pd.DataFrame(ttlist,columns=['secID','secShortName','secCumProfit','secMeanProfit','secStd','baseMeanProfit','baseStd'])
df['SR_P'] = df['secMeanProfit'] / df['secStd'] 
df['SR_B'] = df['baseMeanProfit'] / df['baseStd']
```
