# 马科维茨仓位配置

```python
import numpy as np
import scipy.optimize as sco
import pandas as pd
import matplotlib.pyplot as plt
start = '20190101'
end = '20190712'
universe = ['002157.XSHE','601318.XSHG','000876.XSHE','603806.XSHG','600587.XSHG']
data = pd.DataFrame()
for capitial in universe:
    ret = DataAPI.MktEqudGet(beginDate=start, endDate=end, secID=capitial,pandas='1',\
                            field=['tradeDate', 'secShortName', 'closePrice', 'preClosePrice'])
    ret['returns'] = ret.closePrice/ret.preClosePrice-1
    data = pd.concat([data,ret], axis=0)
data = pd.crosstab(data.tradeDate,data.secShortName,data.returns,aggfunc=max)
data = data.fillna(0)
data.corr()

#-------根据数据求有效边界，ORP，MVP-----------
exp_ret = data.mean() * 252
cov_mat = data.cov() * 252
risk_free = 0.04
sec_num = len(exp_ret)
def random_combination():
    sec_num = len(exp_ret)
    random_rets = []
    random_stds = []
    for i in range(4000):
        weights = np.random.random(sec_num)
        weights = weights / np.sum(weights)
        ret, std, _ = statistic(weights)
        random_rets.append(ret)
        random_stds.append(std)
    return(random_rets, random_stds)

def statistic(weights, exp_ret=exp_ret, cov_mat=cov_mat):
    sec_num = len(exp_ret)
    weights = np.asmatrix(weights).T
    exp_ret = np.asmatrix(exp_ret).T
    cov_mat = np.asmatrix(cov_mat)
    ret = float(weights.T * exp_ret)
    std = float(np.sqrt(weights.T * cov_mat * weights))
    sharpe = (ret - risk_free) / std
    return(ret, std, sharpe)

def efficiency_frontier(exp_ret=exp_ret, cov_mat=cov_mat):
    sec_num = len(exp_ret)
    min_ret = exp_ret.min()
    max_ret = exp_ret.max()
    rets = []
    stds = []
    for ret_level in np.linspace(min_ret, max_ret, 200):
        P = matrix(2 * cov_mat.values)
        q = matrix(np.zeros(sec_num))
        G = matrix(-np.eye(sec_num))
        h = matrix(np.zeros(sec_num))
        A = matrix(np.array([np.ones(sec_num),exp_ret.values]))
        b = matrix([1.0,ret_level])
        solvers.options['show_progress'] = False
        sol = solvers.qp(P,q,G,h,A,b)
        rets.append(ret_level)
        stds.append(np.sqrt(sol['primal objective']))
    return(rets, stds)
    
def min_sharpe(weights):
    return(-statistic(weights)[2])
from cvxopt import solvers, matrix

cons = {'type':'eq', 'fun':lambda x:np.sum(x) - 1.0}
bnds = sec_num * ((0,1),)
opts = sco.minimize(min_sharpe, sec_num*[1.0/sec_num], method='SLSQP', bounds=bnds, constraints=cons)
orp_ret, orp_std, orp_sharpe = statistic(opts['x'])
rets, stds= efficiency_frontier()
random_rets, random_stds=random_combination()

fig = plt.figure(figsize=(14,8))
ax = fig.add_subplot(111)
plt.plot(0,risk_free,'k*',markersize=10)              
ax.plot(stds,rets)                                   
ax.plot(orp_std,orp_ret,'r*',markersize=10)           
plt.scatter(random_stds, random_rets, c=np.array(random_rets)/np.array(random_stds))  
plt.colorbar(label='efficient')
plt.plot((0,orp_std),(risk_free,orp_ret),c='r')
ax.set_xlim(0,0.5)

pd.DataFrame(index=exp_ret.index, data=opts['x'].round(4), columns=['weight'])
```
