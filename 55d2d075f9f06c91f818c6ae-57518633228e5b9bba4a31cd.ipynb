    Type:       function
    Definition: set_universe(symbol=None, date=None)
    Docstring:
    获取预设证券代码列表

    :param str symbol: 预设证券代码列表的名字，Quartz支持以下七个指数：上证50(SH50)、上证180(SH180)、沪深300(HS300)、中证500(ZZ500)、创业板(CYB)、中小板(ZXB)、全A股(A)，可以获得这七个指数的最新成分股列表；同时还支持自定义指数代码，可以获取对应指数的最新成分股列表，如"000001.ZICN"

    :param str/datetime date: 预设证券代码列表日期，仅在上证50、上证180、沪深300、中证500或自定义指数代码时有效，表示获取该日的该指数成分股；默认为最近一个交易日

    :return: 预设证券代码列表

    :rtype: list
