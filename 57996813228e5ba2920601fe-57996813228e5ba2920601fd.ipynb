## 新的模型的策略思路

### 买入条件
   - 1、选取之前15天的MA，在5天前金叉，再前面也是MA10 < MA20
   - 2、上一交易日MA20比上再前一日交易日MA20的比例大于1

实测出来的结果`斜率为1的2015年收益率是95%`，`1.005是106.5%`而`1.01的年华收益率是56%`。担心1.005存在过拟合，所以还是先用1

### 卖出条件
   - 低于成本价95%
   - 最大回撤10%
