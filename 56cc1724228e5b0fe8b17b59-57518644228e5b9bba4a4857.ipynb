首先逻辑就是，判断涨跌趋势：我选择了用均线来做，长期短期均线来判断买卖点。

其次因为放量缩量上涨下跌的这一些市场普遍认同，所以想法就是：利用成交金额来放大或者缩小来反映到当日的成交价格，然后利用成交价格的均线来做择时。

优化参数
-----
两个参数：长短期均线的日期分别为多少。
长期范围：[60,65,..100]
短期范围：[5,6,7,8...,20]
目标：少亏多赚以及正确率

```python
import numpy as np
#测试时间段
example=DataAPI.MktIdxdGet(ticker='000001',beginDate='20050101',endDate='20131231',field=[],pandas="1")
example=example.set_index(example['tradeDate'].values)
ex_close=example['closeIndex']
ex_turnover=example['turnoverValue']
move_ave=pd.rolling_mean(ex_turnover,60)
new_close=ex_close*(ex_turnover/move_ave)
longs=range(60,100,5)
short=range(10,21)
record=np.zeros((5,len(longs)*len(short)))
#买卖信号
def getdataframe(li,lj,index,time,new_close,ex_close):
    example=pd.DataFrame(np.zeros((len(time),4)),columns=['order','close','return','judge'],index=time)
    example['close']=example['close']+ex_close
    short_close=pd.rolling_mean(new_close,li)
    long_close=pd.rolling_mean(new_close,lj)
    i=1
    while i <len(long_close):
        if short_close.ix[i-1]<=long_close.ix[i-1] and short_close.ix[i]>long_close.ix[i]: #买入
            example['order'].ix[i]=1
        if short_close.ix[i-1]>=long_close.ix[i-1] and short_close.ix[i]<long_close.ix[i]: #卖出
            example['order'].ix[i]=-1
        i=i+1
    result=example.ix[example['order']!=0]
    for j in range(len(result)-1):
        result['return'].values[j]=(result['close'].values[j+1]-result['close'].values[j])/result['close'].values[j]   
    result['judge']=result['return']*result['order']
    return result
#统计
l=[]
s=[]
winper=[]
signal_positive=[]
signal_negative=[]
positive_mean_return=[]
negative_mean_return=[]
for i in longs:
    for j in short:
        l.append(i)
        s.append(j)
        data=getdataframe(j,i,ex_close.values,example['tradeDate'].values,new_close,ex_close)
        data3=data.ix[data['judge']>=0.05] #判断对且收益超过5%的行情
        data4=data.ix[data['judge']<-0.05] #判断错且损失超过-5%的行情
        winper.append(float(len(data3))/(float(len(data4))+float(len(data3))))
        signal_positive.append(np.mean(data3['judge'].values))
        signal_negative.append(np.mean(data4['judge'].values) )                   
record[0]=l
record[1]=s
record[2]=winper
record[3]=signal_positive
record[4]=signal_negative
record=record.T
```

```python
Final=pd.DataFrame(record,columns=['long','short','winper','pos_num','neg_num'])
Final=Final.sort(columns='winper')
Final.tail()
```

综合行情胜率以及判断正确时所获利大化，以及判断错误时所受到的损失小化。

选择60，12

首先先选胜率较高的吧

然后每笔判断买信号出现时，平均有拿到16.2%的收益，同时判断错误出现超过-5%的信号时，平均却有9.9%的损失。
