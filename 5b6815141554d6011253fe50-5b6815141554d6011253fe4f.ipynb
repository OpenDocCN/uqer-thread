```python
import numpy as np
import pandas as pd
b=DataAPI.EquIndustryGet(secID=u"",ticker=u"",industryVersionCD=u"010303",intoDate=u"20180302",field=['secID', 'ticker', 'industryName1'],pandas="1")
c = DataAPI.MktStockFactorsOneDayGet(tradeDate=u"20180302",secID=u"",ticker=u"",field=['ticker', 'PB'],pandas="1")
d = pd.merge(b, c, on='ticker', how='left')

d.iloc[1, 2] = 'NaN'
d['PB'] = d['PB'].groupby(d['industryName1']).apply(lambda x: x.fillna(x.mean()))
d
```
