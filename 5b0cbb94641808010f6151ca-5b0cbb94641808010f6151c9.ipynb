# 短期反转和同期换手率之间的交互作用

## 介绍

本文利用非独立双排序的方法，考察单月反转效应的强弱随着换手率高低的变化。结果显示单月反转现象随着同期换手率提升而加强，而随着同期换手率的减少，反转效应越来越弱，在换手率最低的十分位分组中甚至出现了微弱的动量迹象。本文的主要思路和做法借鉴自《Short-term Momentum》 [Medhat et al.[2018]](https://papers.ssrn.com/sol3/papers.cfm?abstract_id=3150525)


有趣的是这篇文章的结果与本文在A股中的发现相反，在美股和其他22个国际上发达的股票市场（包括港股）的实证结果表明，高换手率的股票表现为短期动量，而低换手率的股票表现为短期反转。

## 数据和主要结果
### 单月反转效应

这里使用uqer DataAPI月度行情数据中的单月换手率和单月收益率，通过非独立双排序先根据换手率高低排成10大组，然后在每一大组中再根据近一个月的换手率再排成8个小组,然后统计每组中下个月的平均收益率（相当于等权组合），结果如下：
![return_on_turnover_8x10.png](http://storage-uqer.datayes.com/5ab0ad1aff59930135a72b7c/b0691ed8-54ba-11e8-b7cc-0242ac140002)

_其中y轴代表换手率从低到高的十个大组，x轴代表单月收益率从低到高的分组，组号越大代表该组别对应因子值越大。_


从上图可以看出：

0. 整体来看，高换手、高单月收益率的股票是雷区（图中上三角区域），除此之外换手率和单月收益率对未来收益率的影响不明显（图中下三角的“高原”区域）；

1. 高换手的股票中短期反转效应明显，而随着换手率降低单月收益率对未来收益的区分度越来越小，这主要是由于高收益的股票预测未来下跌的作用减弱。

2. 未来收益最高和最低的分组都出现在短期表现最好的类别中，分别对应于换手率最低和最高的组别。换句话说，同期换手率的高低决定了过去一个月的上涨是否可以持续。这让我想起了《投资中最简单的事》中说的“估值决定了股票能够上涨的空间，流动性则决定了股市涨跌的时间。”

3. 在换手率最小的分组中有动量现象，亦即单月上涨的股票下个月依然上涨，这个结论即使在剔除（8，1）这个收益异常高的分组后依然成立。

4. 此热力图的结果和流动性/反转因子的单因子分组表现一致。相当于将一个二维随机变量的联合概率分布转换成边缘分布。

接下来观察不同规模公司的单月反转现象，这里将股票分为沪深300、中证500和其他三个类别，可以看出在反转效应在小市值的股票中更为显著。

![return_on_sizeidx_8x3](http://storage-uqer.datayes.com/5ab0ad1aff59930135a72b7c/e7bdd7e8-54ba-11e8-b7cc-0242ac140002)

### 现象的解释

在个人投资者成交占主导的市场，高换手的股票意味着更多个人投资者的交易，这些交易往往不是基于基本面的。而股价近期的大涨/大跌则意味着拥挤的交易已经推动股价脱离基本面，故未来将出现反转。过去换手率越大噪音交易的越频繁。对于换手率极低的股票，“流动性决定股票上涨的时间”，低流动性使得股价需要更多时间来反映预期的变化，另一方面低流动性股票的市场关注度低信息的扩散更慢。

至于公司规模对反转效应的影响，我认为可能的一个原因也是市场关注度的差别，另外就是由于股指期货的存在，沪深300和中证500有更多的统计套利者维持这部分市场的有效性，而小盘股除了寄希望于长期投资者的进入似乎没有其他的机制。

最后，市场情绪类因子基本可以用噪音交易的来源，以及保持价格偏离客观价值的市场结构来解释。但基于投资者行为的解释若缺少直接证据，充其量只是一个能够自圆其说的猜测，多少有些牵强附会。

## 月末时间段对单月收益率因子的影响

这里将过去一个月的统计时段按月末前五个交易日为界切分为前后两个时间段。分别研究这两段时间收益率对下个月收益率的预测作用。结果如下：

### 月初前16交易日

![return_on_turnover_8x10.png](http://storage-uqer.datayes.com/5ab0ad1aff59930135a72b7c/b0691ed8-54ba-11e8-b7cc-0242ac140002)

Figure: 单月反转因子 VS 单月换手率

![return16_on_turnover_8x10](http://storage-uqer.datayes.com/5ab0ad1aff59930135a72b7c/f79a354e-54ba-11e8-a9af-0242ac140002)

Figure: 前16交易日反转因子 VS 前16交易日换手率

从收益率数值的spread上可以看出，剔除最后5交易日后相同换手率分组(比如换手率第10组)的反转效应减弱了，下文对公司规模作双排序也有相同的结论，造成这一现象的主要原因是信号的decay。注意到原文中剔除月末三个交易日后动量效应有增强，而在A股这一现象不明显。

![return_on_sizeidx_8x3](http://storage-uqer.datayes.com/5ab0ad1aff59930135a72b7c/e7bdd7e8-54ba-11e8-b7cc-0242ac140002)

Figure: 单月反转因子 VS 规模分类

![return16_on_sizeidx_8x3](http://storage-uqer.datayes.com/5ab0ad1aff59930135a72b7c/249dffb2-54bb-11e8-a9af-0242ac140002)

Figure: 前16交易日反转因子 VS 规模分类

原文对美股的解释是月末的换手相比其他时间段含有更多基于机构流动性需要的uninformed trading，所以无法产生短期动量。在A股，对于前16交易日的收益率剔除月末5天后的反转效应减弱了，而动量效应也有所减弱，我们分析是信号decay的原因。

### 月末前5交易日

如果用月末前5交易日的收益率来预测下个月的收益率，月末前5交易日收益率接近市场中位数的股票下月表现最好。而原文在美股的结果则表明这段时间对反转效应有增强的作用。

![return5_on_turnover_8x10](http://storage-uqer.datayes.com/5ab0ad1aff59930135a72b7c/344c964e-54bb-11e8-a9af-0242ac140002)

Figure: 后5交易日反转因子 VS 后5交易日换手率

![return5_on_sizeidx_8x3](http://storage-uqer.datayes.com/5ab0ad1aff59930135a72b7c/443a3a3e-54bb-11e8-a9af-0242ac140002)

Figure: 后5交易日反转因子 VS 规模分类

造成结果不同的原因可能是：

1. 原文的因子是最后3个交易日的收益率，且建议使用最后一个交易日的收益率作为因子。而本文因为便利性的原因用的是最后5个交易日的收益率。
2. 上文提到的流动性需要之外的其他原因造成在月末下跌股票中体现出较强的动量效应。

尽管这里对A股月末收益率横截面的观察没有发现反转的证据，但时间序列上却存在月末的反转，见社区里的另一篇帖子 [《A股换月效应的多角度解析》](https://uqer.datayes.com/community/share/5ab0ae6cff599305a0a729cf)

## 总结

《Short-term Momentum》这篇文章中作者构建了一个基于交易者行为假设的模型，模拟了噪音交易者，基本面交易者以及基本面交易者对于同一新闻因观点分歧导致的不同交易的行为，再通过数学推导证明了随着换手率提高股票将表现出动量，海外市场同时也观察到了这些现象。A股的神奇之处是我们观察到的现象是相反的，虽然我们没有给出A股的交易者行为模型，但也尝试从常识和直觉出发给出一些解释。在月末收益率研究中我没有发现反转效应，后续改进工作可以尝试把统计时间段缩小到月末最后1个交易日。

```python
# 整月的收益率反转

import pandas as pd
import numpy as np
import seaborn as sns

import re
import matplotlib.pyplot as plt
import matplotlib.cm as cm

import statsmodels.api as sm
import statsmodels.formula.api as smf
from functools import partial
import datetime

from scipy.stats.mstats import winsorize, zscore
trading_cal = DataAPI.TradeCalGet(exchangeCD=u"XSHG",beginDate=u"",endDate=u"",field=u"",pandas="1").query('isMonthEnd==1')
trading_cal['calendarDate_timestamp'] = trading_cal.calendarDate.astype(pd.Timestamp)
trading_cal['calendarDate_timestamp'] = pd.to_datetime(trading_cal['calendarDate_timestamp'])

trading_cal = trading_cal[trading_cal['calendarDate_timestamp']>pd.Timestamp('2011-01-01')]
trading_cal.sort_values(by='calendarDate_timestamp',inplace=True)
trading_cal['last_calendarDate_of_next_month'] = trading_cal['calendarDate'].shift(-1)


a = []
for mon_end in trading_cal['calendarDate']:
    dd = DataAPI.MktEqumAdjAfGet(secID=u"",ticker=u"",monthEndDate=mon_end,beginDate=u"",endDate=u"",isOpen=u"",field=u"",pandas="1")
    dd['last_calendarDate_of_month'] = mon_end

    a.append(dd)

df = pd.concat(a)
df1 = df


a = []
for mon_end in trading_cal['calendarDate']:
    dd = DataAPI.IdxConsGet(secID=u"",ticker=u"000300",isNew=u"",intoDate=mon_end,field=u"",pandas="1")
    dd['last_calendarDate_of_month'] = mon_end

    a.append(dd)

hs300 = pd.concat(a)

a = []
for mon_end in trading_cal['calendarDate']:
    dd = DataAPI.IdxConsGet(secID=u"",ticker=u"000905",isNew=u"",intoDate=mon_end,field=u"",pandas="1")
    dd['last_calendarDate_of_month'] = mon_end

    a.append(dd)

zz500 = pd.concat(a)


# 过滤停牌
df1 = df1[df1['tradeDays']!=0]
df1['turnoverRate'].replace(0,np.nan,inplace=True)
df1['return'].replace(0,np.nan,inplace=True)
df1['return'].replace(-1,np.nan,inplace=True)

df1['avgTurnoverRate'].replace(0,np.nan,inplace=True)

def qcut(*args,**kwargs):
    try:
        a = pd.qcut(*args,**kwargs)
        return a
    except:
        print args


df2 = df1.merge(trading_cal,right_on='calendarDate',left_on='last_calendarDate_of_month',)
df3 = df2.merge(df1,left_on=['secID','last_calendarDate_of_next_month'],right_on=['secID','last_calendarDate_of_month'],suffixes=('_t','_t+1'))
df3['turnoverRate_disc_t'] = df3.groupby(by='last_calendarDate_of_month_t')['turnoverRate_t'].apply(qcut,10,labels=range(1,11))

df3['avgTurnoverRate_disc_t'] = df3.groupby(by='last_calendarDate_of_month_t')['avgTurnoverRate_t'].apply(qcut,10,labels=range(1,11))
df3['return_disc_t'] = df3.groupby(by=['last_calendarDate_of_month_t','turnoverRate_disc_t'])['return_t'].apply(qcut,8,labels=range(1,9))

aa_p = df3.groupby(by=['turnoverRate_disc_t','return_disc_t'])['return_t+1'].mean()



aa_pp = aa_p.reset_index(drop=False)
aa_pp = aa_pp.pivot(u'turnoverRate_disc_t', u'return_disc_t', u'return_t+1')
aa_pp = aa_pp.loc[aa_pp.index[::-1]]
sns.heatmap(aa_pp, annot=True, fmt=".4f", linewidths=.5)
plt.show()


idx_cons = pd.concat([hs300,zz500])
idx_cons = idx_cons[['last_calendarDate_of_month','consID','secID']]
idx_cons.rename(columns={'secID':'idxID'},inplace=True)

df33 = df3.merge(idx_cons,left_on=['last_calendarDate_of_month_t','secID'],right_on=['last_calendarDate_of_month','consID'],how='outer')
df33['idxID'] = df33['idxID'].fillna('Other')
df33['return_disc2_t'] = df33.groupby(by=['last_calendarDate_of_month_t','idxID'])['return_t'].apply(qcut,8,labels=range(1,9))

aa_p1 = df33.groupby(by=['idxID','return_disc2_t'])['return_t+1'].mean()

aa_pp1 = aa_p1.reset_index(drop=False)
aa_pp1 = aa_pp1.pivot('idxID', 'return_disc2_t', 'return_t+1')
aa_pp1 = aa_pp1.loc[aa_pp1.index[::-1]]
sns.heatmap(aa_pp1, annot=True, fmt=".4f", linewidths=.5)
plt.show()
#

```
