                                                            2.因子IC检验 
IC即信息系数（InformationCoefficient），表示所选因子值与下期收益率的截面相关系数，通过 IC 值可以判断因子值对下期收益率的预测能力。信息系数的绝对值越大，该因子越有效。IC为负表示因子值越小越好，IC为正表示因子值越大越好。
IC的计算方法是：计算全部股票在调仓周期期初排名和调仓周期期末收益排名的线性相关度（Correlation）。IC绝对值越大的因子，有效性越强。
IC的计算方式有两种：Normal IC 和 Rank IC
Normal IC
IC（Informationcoefficient 信息系数）的定义：t期（这里的期一般指的是调仓周期）的因子载荷（因子值）对t+1期的收益预测值和实际收益之间的相关系数。
IC=corr(f，r)
IC：因子在该期的IC值
f：t期因子对T+1期收益率的预测值
r：t+1期股票实际收益率
要计算IC就需要知道因子对下期收益率的计算。很多文献中是直接用t期的因子值来当作这个对t + 1期的收益预测值，也就是说，直接计算t期因子值与t + 1期的收益率之间的相关系数，作为IC值。
Rank IC
因为normal IC有一个前提条件，就是数据要服从正态分布，现实往往不理想，所以实际中更多人采用rank IC（秩相关系数）来判断因子的有效性。
Rank IC，即某时点某因子在全部股票暴露值排名与其下期回报排名的截面相关系数，它与IC的区别就是将因子的具体值以及收益的具体值，都转换为了所对应的数值在其截面上的排序名次。然后通过计算排序值的相关系数（秩相关系数），得到Rank IC。两者分别对应Pearson 或者 Spearman 相关系数（使用范围更广）。
因此我们采用spearman相关系数计算IC。

计算IR:
IR即信息比率（InformationRatio），是超额收益的均值与标准差之比，代表因子获取稳定Alpha的能力。
通常整个回测时段由多个调仓周期组成，每一个周期都会计算出一个不同的IC值，IR等于多个调仓周期的IC均值除以这些IC的标准方差
                               IR=IC的多周期均值/IC的多周期标准差
所以IR兼顾了因子的有效性（由IC代表）和因子稳定性（由IC的标准方差的倒数代表）
实际中对于多年的IC均值，当均值绝对值 > 0.05时，就可以视为有效因子，当I C > 0.1，就可以认为是很好的阿尔法因子了。当IC均值接近0，可视为无效因子。
IR大于0.5时，因子稳定获取超额收益的能力较强。

                                                                3.因子分组回测检验
上面完成了各因子的IC检验，接下来分析分组回测。
回测过程：按照因子值对股票进行排序，然后进行十分位回测（按照因子值将股票分位10组）。每个月最后一个交易日根据因子值对股票进行排序，下个月第一个交易日重新建仓，卖出不在相应分位段的股票，买入重新分入该分位段的股票，直到回测结束。
因为因子太多，没法都进行回测（一个因子大概要跑半小时的时间，如果强行对所有因子回测可能需要跑好几天，并且内寸大小不支持跑完所有因子），那么我们选取IC检验最好的也就是IC绝对值最大的五个因子（CTOP,DDNSR,EPS,HSIGMA,VOL20（这里换手率因子都是有效性极高的因子，因此只选取其中最好的VOL20，五个因子跑了三个小时））进行回测得到其多头组合的年化收益，年化超额收益（策略的年化收益减去基准的年化收益），超额最大回撤（累计超额收益的最大回撤），平均换手率以及单调性，并展现这些因子的多头收益表现（相对净值曲线：比较基准为沪深300）。

这里对选取的几个因子做解释：
CTOP：现金流市值比（Cash flow to price）。计算方法：现金流市值比=每股派现（税前）×分红前总股本/总市值。属于价值类因子。
DDNSR：下跌波动（Downside standard deviations ratio），过往12个月中，市场组合日收益为负时，个股日收益标准差和市场组合日收益标准差之比。属于收益和风险类因子。
EPS：股东权益与固定资产比率（Equity fixed assets ratio）。计算方法：股东权益与固定资产比率=股东权益/(固定资产+工程物资+在建工程)。属于质量类因子。
HSIGMA：历史波动（Historical daily sigma），过往12个月中，个股日收益关于市场组合日收益的三阶自回归，市场组合日收益的残差标准差。属于收益和风险类因子。
VOL20：20日平均换手率（Turnover Rate）。属于情绪类因子。

实现：我们利用数据框展示了单个因子的多头年化收益率、多头年化超额收益率、多头超额最大回撤等指标（展示多头组合的选股能力），其中我们用较为创新的方式对单调性进行计算，即测算分组回测的组数序号与对应组合年化收益率的秩相关系数，该绝对数值越接近于1，则因子单调性越显著。最后画出因子每组的相对净值曲线，即该组的净值与基准（沪深300）净值的比较。


```python
import time
from CAL.PyCAL import * #导入python日历模块？

start_time = time.time()

# -------------------------------------------------1、设定回测参数 ------------------------------------------------------

start = '2010-01-05'                       # 回测起始时间
end = '2022-8-31'                         # 回测结束时间
benchmark = 'HS300'                        # 策略参考标准
universe = DynamicUniverse('HS300')        # 证券池，支持股票和基金
capital_base = 10000000                    # 起始资金
freq = 'd'                                 # 策略类型，'d'表示日间策略使用日线回测，'m'表示日内策略使用分钟线回测
refresh_rate = Monthly(1)                  # 调仓频率，表示执行handle_data的时间间隔，若freq = 'd'时间间隔的单位为交易日，若freq = 'm'时间间隔为分钟

# 配置账户信息，支持多资产多账户
accounts = {
    'fantasy_account': AccountConfig(account_type='security', capital_base=10000000)
}

# ------------------------------------------------- 2、读取因子数据 -------------------------------------------------------
factor_data = tfactor_frame[['secID', 'tradeDate', 'CTOP']]#取dataframe里的'secID', 'tradeDate', 'ROE'三列数据
factor_data.rename(columns={"CTOP":'factor'}, inplace=True)#列重命名
factor_data = factor_data.set_index('tradeDate', drop=True)#把'tradeDate'设置为索引
factor_data.dropna(inplace=True)# 默认情况下，axis = 0，即在行上应用，这意味着如果行内的任何值是NA，那么整个行被排除
q_dates = factor_data.index.values#获取因子日期



# ------------------------------------------------3、将因子分组并回测 ---------------------------------------------------------

# 调用quick_backtest方法
# a、回测初始化配置，在代码模式进行策略回测时使用
sim_params = quartz.SimulationParameters(start, end, benchmark, universe, capital_base, refresh_rate=refresh_rate, accounts=accounts)
# b、获取回测行情数据
data = quartz.get_backtest_data(sim_params)
# c、运行结果
results = {}

# d、遍历每一组，进行回测
for quantile_ten in range(1, 11):  
    
    def initialize(context):                   
        pass
    # d1、初始化虚拟账户状态，
    # 每个单位时间(如果按天回测,则每天调用一次,如果按分钟,则每分钟调用一次)调用一次

    def handle_data(context): 
    # d2、策略逻辑实现
        account = context.get_account('fantasy_account')#设定交易账户
        current_universe = context.get_universe('stock', exclude_halt=True)
        #获取当前交易日证券池，获取当前交易日的证券池，是策略初始化参数中universe的子集。在这种配置情况下，context.get_universe()仅体现资产的上市状态，只要资产在策略运行当天处于上市状态，就可通过 context.get_universe()获得，exclude_halt=True去除资产池中的停牌股票
        pre_date = context.previous_date.strftime("%Y%m%d")#获取当前回测日期的前一交易日，并输出指定格式字符串
        if pre_date not in q_dates:            
            return
        # 拿取调仓日前一个交易日的因子，并按照相应分位选择股票
        q = factor_data.ix[pre_date].dropna()#获取因子数据，空值删除
        q = q.set_index('secID', drop=True)#把'secID'设置为索引
        q = q.ix[current_universe]#取出在当前证券池里面的股票，已剔除停盘股票
        q.dropna(inplace=True)
        q_min = q['factor'].quantile((quantile_ten-1)*0.1)
        q_max = q['factor'].quantile(quantile_ten*0.1)
        my_univ = q[(q['factor']>=q_min) & (q['factor']<q_max)].index.values#目标证券池
       # 交易部分
        positions = account.get_positions()
        #获取所有账户持仓，exclude_halt是否移除持仓中停牌的资产，前面已剔除停牌股票
        sell_list = [stk for stk in positions if stk not in my_univ]
        for stk in sell_list:
            account.order_to(stk,0)#如果持仓股票不在当期目标证券池里面，全部卖出
        # 在目标股票池中的，等权买入
        for stk in my_univ:
            account.order_pct_to(stk, 1.0/len(my_univ))#按等比例买入目标证券池里面的股票


    # d3、生成策略对象
    strategy = quartz.TradingStrategy(initialize, handle_data)
    
    # d4、开始回测, 返回说明见：https://uqer.datayes.com/help/api/#策略结果展示
    bt, perf = quartz.quick_backtest(sim_params, strategy, data=data)

    # d5、保存运行结果，1为因子最强组，10为因子最弱组
    # 字典里面嵌套字典
    results[11 - quantile_ten] = { 'annualized_return': perf['annualized_return'], 'turnover': perf['turnover_rate'],'extra_annualized_return': perf['annualized_return']-perf['benchmark_annualized_return'], 'extra_portfolio': perf['cumulative_returns']-perf['benchmark_cumulative_returns'],'bt': bt}      

    print ('backtesting for group %s..................................' % str(quantile_ten)),
```

```python
import seaborn as sns#数据可视化-seaborn
import matplotlib.pyplot as plt
sns.set_style('white')
#set_style( )是用来设置主题的，Seaborn有五个预设好的主题： darkgrid , whitegrid , dark , white ,和 ticks  默认： darkgrid
#sns.set(style="white", palette="muted", color_codes=True)     #set( )设置主题，调色板更常用  

fig = plt.figure(figsize=(10,8))#图像的大小设置
fig.set_tight_layout(True)#调整图像边缘及图像间的空白间隔，图像外部边缘的调整可以使用plt.tight_layout()进行自动控制
ax1 = fig.add_subplot(211)#第一个参数子图总行数，第二个参数子图总列数，第三个参数子图位置
ax1.grid()#面向对象的方式
# True 显示网格  
# linestyle 设置线显示的类型(一共四种)  
# color 设置网格的颜色  
# linewidth 设置网格的宽度   
#plt.grid(True, linestyle = "-.", color = "r", linewidth = "3")  


for qt in results:
    bt = results[qt]['bt']
    data = bt[[u'tradeDate',u'portfolio_value',u'benchmark_return']]
    data['portfolio_return'] = data.portfolio_value/data.portfolio_value.shift(1) - 1.0   # 总头寸每日回报率
    data['portfolio_return'].ix[0] = data['portfolio_value'].ix[0]/	10000000.0 - 1.0
    data['excess_return'] = data.portfolio_return - data.benchmark_return                 # 总头寸每日超额回报率
    data['portfolio'] = data.portfolio_return + 1.0     
    data['portfolio'] = data.portfolio.cumprod()          # 总头寸的净值序列
    data['benchmark'] = data.benchmark_return + 1.0
    data['benchmark'] = data.benchmark.cumprod()          # benchmark的净值序列
    data['relative_value'] = data.portfolio/data.benchmark #相对净值
    results[qt]['extra_portfolio']=results[qt]['extra_portfolio']+1
    A = results[qt]['extra_portfolio']
    results[qt]['extra_max_drawdown'] = max([1 - v/max(1, max(A[:i+1])) for i,v in enumerate(A)])
    ax1.plot(data['tradeDate'], data[['relative_value']], label=str(qt))

    

ax1.legend(loc=0)
ax1.set_ylabel(u"净值", fontproperties=font, fontsize=16)
ax1.set_title(u"因子不同十分位数分组选股净值走势", fontproperties=font, fontsize=16)

# results 转换为 DataFrame
#把字典嵌套字典转化为dataframe，外层字典keys作为columns，内层字典keys作为index
results_pd = pd.DataFrame(results).T.sort_index()

results_pd = results_pd[[u'annualized_return', u'extra_max_drawdown',  u'turnover' ,u'extra_annualized_return']]

results_pd.columns = [u'年化收益',u'超额最大回撤',u'换手率',u'超额年化收益']
results_pd.index.name = u'十分位组别'
results_pd
```

```python
import time
from CAL.PyCAL import * #导入python日历模块？

start_time = time.time()

# -------------------------------------------------1、设定回测参数 ------------------------------------------------------

start = '2010-01-05'                       # 回测起始时间
end = '2022-8-31'                         # 回测结束时间
benchmark = 'HS300'                        # 策略参考标准
universe = DynamicUniverse('HS300')        # 证券池，支持股票和基金
capital_base = 10000000                    # 起始资金
freq = 'd'                                 # 策略类型，'d'表示日间策略使用日线回测，'m'表示日内策略使用分钟线回测
refresh_rate = Monthly(1)                  # 调仓频率，表示执行handle_data的时间间隔，若freq = 'd'时间间隔的单位为交易日，若freq = 'm'时间间隔为分钟

# 配置账户信息，支持多资产多账户
accounts = {
    'fantasy_account': AccountConfig(account_type='security', capital_base=10000000)
}

# ------------------------------------------------- 2、读取因子数据 -------------------------------------------------------
factor_data = tfactor_frame[['secID', 'tradeDate', 'EPS']]#取dataframe里的'secID', 'tradeDate', 'ROE'三列数据
factor_data.rename(columns={"EPS":'factor'}, inplace=True)#列重命名
factor_data = factor_data.set_index('tradeDate', drop=True)#把'tradeDate'设置为索引
factor_data.dropna(inplace=True)# 默认情况下，axis = 0，即在行上应用，这意味着如果行内的任何值是NA，那么整个行被排除
q_dates = factor_data.index.values#获取因子日期



# ------------------------------------------------3、将因子分组并回测 ---------------------------------------------------------

# 调用quick_backtest方法
# a、回测初始化配置，在代码模式进行策略回测时使用
sim_params = quartz.SimulationParameters(start, end, benchmark, universe, capital_base, refresh_rate=refresh_rate, accounts=accounts)
# b、获取回测行情数据
data = quartz.get_backtest_data(sim_params)
# c、运行结果
results = {}

# d、遍历每一组，进行回测
for quantile_ten in range(1, 11):  
    
    def initialize(context):                   
        pass
    # d1、初始化虚拟账户状态，
    # 每个单位时间(如果按天回测,则每天调用一次,如果按分钟,则每分钟调用一次)调用一次

    def handle_data(context): 
    # d2、策略逻辑实现
        account = context.get_account('fantasy_account')#设定交易账户
        current_universe = context.get_universe('stock', exclude_halt=True)
        #获取当前交易日证券池，获取当前交易日的证券池，是策略初始化参数中universe的子集。在这种配置情况下，context.get_universe()仅体现资产的上市状态，只要资产在策略运行当天处于上市状态，就可通过 context.get_universe()获得，exclude_halt=True去除资产池中的停牌股票
        pre_date = context.previous_date.strftime("%Y%m%d")#获取当前回测日期的前一交易日，并输出指定格式字符串
        if pre_date not in q_dates:            
            return
        # 拿取调仓日前一个交易日的因子，并按照相应分位选择股票
        q = factor_data.ix[pre_date].dropna()#获取因子数据，空值删除
        q = q.set_index('secID', drop=True)#把'secID'设置为索引
        q = q.ix[current_universe]#取出在当前证券池里面的股票，已剔除停盘股票
        q.dropna(inplace=True)
        q_min = q['factor'].quantile((quantile_ten-1)*0.1)
        q_max = q['factor'].quantile(quantile_ten*0.1)
        my_univ = q[(q['factor']>=q_min) & (q['factor']<q_max)].index.values#目标证券池
       # 交易部分
        positions = account.get_positions()
        #获取所有账户持仓，exclude_halt是否移除持仓中停牌的资产，前面已剔除停牌股票
        sell_list = [stk for stk in positions if stk not in my_univ]
        for stk in sell_list:
            account.order_to(stk,0)#如果持仓股票不在当期目标证券池里面，全部卖出
        # 在目标股票池中的，等权买入
        for stk in my_univ:
            account.order_pct_to(stk, 1.0/len(my_univ))#按等比例买入目标证券池里面的股票


    # d3、生成策略对象
    strategy = quartz.TradingStrategy(initialize, handle_data)
    
    # d4、开始回测, 返回说明见：https://uqer.datayes.com/help/api/#策略结果展示
    bt, perf = quartz.quick_backtest(sim_params, strategy, data=data)

    # d5、保存运行结果，1为因子最强组，10为因子最弱组
    # 字典里面嵌套字典
    results[11 - quantile_ten] = { 'annualized_return': perf['annualized_return'], 'turnover': perf['turnover_rate'],'extra_annualized_return': perf['annualized_return']-perf['benchmark_annualized_return'], 'extra_portfolio': perf['cumulative_returns']-perf['benchmark_cumulative_returns'],'bt': bt}      

    print ('backtesting for group %s..................................' % str(quantile_ten)),
    
    
import seaborn as sns#数据可视化-seaborn
import matplotlib.pyplot as plt
sns.set_style('white')
#set_style( )是用来设置主题的，Seaborn有五个预设好的主题： darkgrid , whitegrid , dark , white ,和 ticks  默认： darkgrid
#sns.set(style="white", palette="muted", color_codes=True)     #set( )设置主题，调色板更常用  

fig = plt.figure(figsize=(10,8))#图像的大小设置
fig.set_tight_layout(True)#调整图像边缘及图像间的空白间隔，图像外部边缘的调整可以使用plt.tight_layout()进行自动控制
ax1 = fig.add_subplot(211)#第一个参数子图总行数，第二个参数子图总列数，第三个参数子图位置
ax1.grid()#面向对象的方式
# True 显示网格  
# linestyle 设置线显示的类型(一共四种)  
# color 设置网格的颜色  
# linewidth 设置网格的宽度   
#plt.grid(True, linestyle = "-.", color = "r", linewidth = "3")  

for qt in results:
    bt = results[qt]['bt']
    data = bt[[u'tradeDate',u'portfolio_value',u'benchmark_return']]
    data['portfolio_return'] = data.portfolio_value/data.portfolio_value.shift(1) - 1.0   # 总头寸每日回报率
    data['portfolio_return'].ix[0] = data['portfolio_value'].ix[0]/	10000000.0 - 1.0
    data['excess_return'] = data.portfolio_return - data.benchmark_return                 # 总头寸每日超额回报率
    data['portfolio'] = data.portfolio_return + 1.0     
    data['portfolio'] = data.portfolio.cumprod()          # 总头寸的净值序列
    data['benchmark'] = data.benchmark_return + 1.0
    data['benchmark'] = data.benchmark.cumprod()          # benchmark的净值序列
    data['relative_value'] = data.portfolio/data.benchmark #相对净值
    results[qt]['extra_portfolio']=results[qt]['extra_portfolio']+1
    A = results[qt]['extra_portfolio']
    results[qt]['extra_max_drawdown'] = max([1 - v/max(1, max(A[:i+1])) for i,v in enumerate(A)])
    ax1.plot(data['tradeDate'], data[['relative_value']], label=str(qt))

    

ax1.legend(loc=0)
ax1.set_ylabel(u"净值", fontproperties=font, fontsize=16)
ax1.set_title(u"因子不同十分位数分组选股净值走势", fontproperties=font, fontsize=16)

# results 转换为 DataFrame
#把字典嵌套字典转化为dataframe，外层字典keys作为columns，内层字典keys作为index
results_pd = pd.DataFrame(results).T.sort_index()

results_pd = results_pd[[ u'annualized_return', u'extra_max_drawdown',  u'turnover' ,u'extra_annualized_return']]

results_pd.columns = [u'年化收益',u'超额最大回撤',u'换手率',u'超额年化收益']
results_pd.index.name = u'十分位组别'
results_pd
```

```python
import time
from CAL.PyCAL import * #导入python日历模块？

start_time = time.time()

# -------------------------------------------------1、设定回测参数 ------------------------------------------------------

start = '2010-01-05'                       # 回测起始时间
end = '2022-8-31'                         # 回测结束时间
benchmark = 'HS300'                        # 策略参考标准
universe = DynamicUniverse('HS300')        # 证券池，支持股票和基金
capital_base = 10000000                    # 起始资金
freq = 'd'                                 # 策略类型，'d'表示日间策略使用日线回测，'m'表示日内策略使用分钟线回测
refresh_rate = Monthly(1)                  # 调仓频率，表示执行handle_data的时间间隔，若freq = 'd'时间间隔的单位为交易日，若freq = 'm'时间间隔为分钟

# 配置账户信息，支持多资产多账户
accounts = {
    'fantasy_account': AccountConfig(account_type='security', capital_base=10000000)
}

# ------------------------------------------------- 2、读取因子数据 -------------------------------------------------------
factor_data = tfactor_frame[['secID', 'tradeDate', 'VOL20']]#取dataframe里的'secID', 'tradeDate', 'ROE'三列数据
factor_data.rename(columns={"VOL20":'factor'}, inplace=True)#列重命名
factor_data = factor_data.set_index('tradeDate', drop=True)#把'tradeDate'设置为索引
factor_data.dropna(inplace=True)# 默认情况下，axis = 0，即在行上应用，这意味着如果行内的任何值是NA，那么整个行被排除
q_dates = factor_data.index.values#获取因子日期



# ------------------------------------------------3、将因子分组并回测 ---------------------------------------------------------

# 调用quick_backtest方法
# a、回测初始化配置，在代码模式进行策略回测时使用
sim_params = quartz.SimulationParameters(start, end, benchmark, universe, capital_base, refresh_rate=refresh_rate, accounts=accounts)
# b、获取回测行情数据
data = quartz.get_backtest_data(sim_params)
# c、运行结果
results = {}

# d、遍历每一组，进行回测
for quantile_ten in range(1, 11):  
    
    def initialize(context):                   
        pass
    # d1、初始化虚拟账户状态，
    # 每个单位时间(如果按天回测,则每天调用一次,如果按分钟,则每分钟调用一次)调用一次

    def handle_data(context): 
    # d2、策略逻辑实现
        account = context.get_account('fantasy_account')#设定交易账户
        current_universe = context.get_universe('stock', exclude_halt=True)
        #获取当前交易日证券池，获取当前交易日的证券池，是策略初始化参数中universe的子集。在这种配置情况下，context.get_universe()仅体现资产的上市状态，只要资产在策略运行当天处于上市状态，就可通过 context.get_universe()获得，exclude_halt=True去除资产池中的停牌股票
        pre_date = context.previous_date.strftime("%Y%m%d")#获取当前回测日期的前一交易日，并输出指定格式字符串
        if pre_date not in q_dates:            
            return
        # 拿取调仓日前一个交易日的因子，并按照相应分位选择股票
        q = factor_data.ix[pre_date].dropna()#获取因子数据，空值删除
        q = q.set_index('secID', drop=True)#把'secID'设置为索引
        q = q.ix[current_universe]#取出在当前证券池里面的股票，已剔除停盘股票
        q.dropna(inplace=True)
        q_min = q['factor'].quantile((quantile_ten-1)*0.1)
        q_max = q['factor'].quantile(quantile_ten*0.1)
        my_univ = q[(q['factor']>=q_min) & (q['factor']<q_max)].index.values#目标证券池
       # 交易部分
        positions = account.get_positions()
        #获取所有账户持仓，exclude_halt是否移除持仓中停牌的资产，前面已剔除停牌股票
        sell_list = [stk for stk in positions if stk not in my_univ]
        for stk in sell_list:
            account.order_to(stk,0)#如果持仓股票不在当期目标证券池里面，全部卖出
        # 在目标股票池中的，等权买入
        for stk in my_univ:
            account.order_pct_to(stk, 1.0/len(my_univ))#按等比例买入目标证券池里面的股票


    # d3、生成策略对象
    strategy = quartz.TradingStrategy(initialize, handle_data)
    
    # d4、开始回测, 返回说明见：https://uqer.datayes.com/help/api/#策略结果展示
    bt, perf = quartz.quick_backtest(sim_params, strategy, data=data)

    # d5、保存运行结果，1为因子最强组，10为因子最弱组
    # 字典里面嵌套字典
    results[11 - quantile_ten] = { 'annualized_return': perf['annualized_return'], 'turnover': perf['turnover_rate'],'extra_annualized_return': perf['annualized_return']-perf['benchmark_annualized_return'], 'extra_portfolio': perf['cumulative_returns']-perf['benchmark_cumulative_returns'],'bt': bt}      

    print ('backtesting for group %s..................................' % str(quantile_ten)),
    
    
import seaborn as sns#数据可视化-seaborn
import matplotlib.pyplot as plt
sns.set_style('white')
#set_style( )是用来设置主题的，Seaborn有五个预设好的主题： darkgrid , whitegrid , dark , white ,和 ticks  默认： darkgrid
#sns.set(style="white", palette="muted", color_codes=True)     #set( )设置主题，调色板更常用  

fig = plt.figure(figsize=(10,8))#图像的大小设置
fig.set_tight_layout(True)#调整图像边缘及图像间的空白间隔，图像外部边缘的调整可以使用plt.tight_layout()进行自动控制
ax1 = fig.add_subplot(211)#第一个参数子图总行数，第二个参数子图总列数，第三个参数子图位置
ax1.grid()#面向对象的方式
# True 显示网格  
# linestyle 设置线显示的类型(一共四种)  
# color 设置网格的颜色  
# linewidth 设置网格的宽度   
#plt.grid(True, linestyle = "-.", color = "r", linewidth = "3")  

for qt in results:
    bt = results[qt]['bt']
    data = bt[[u'tradeDate',u'portfolio_value',u'benchmark_return']]
    data['portfolio_return'] = data.portfolio_value/data.portfolio_value.shift(1) - 1.0   # 总头寸每日回报率
    data['portfolio_return'].ix[0] = data['portfolio_value'].ix[0]/	10000000.0 - 1.0
    data['excess_return'] = data.portfolio_return - data.benchmark_return                 # 总头寸每日超额回报率
    data['portfolio'] = data.portfolio_return + 1.0     
    data['portfolio'] = data.portfolio.cumprod()          # 总头寸的净值序列
    data['benchmark'] = data.benchmark_return + 1.0
    data['benchmark'] = data.benchmark.cumprod()          # benchmark的净值序列
    data['relative_value'] = data.portfolio/data.benchmark #相对净值
    results[qt]['extra_portfolio']=results[qt]['extra_portfolio']+1
    A = results[qt]['extra_portfolio']
    results[qt]['extra_max_drawdown'] = max([1 - v/max(1, max(A[:i+1])) for i,v in enumerate(A)])
    ax1.plot(data['tradeDate'], data[['relative_value']], label=str(qt))

    

ax1.legend(loc=0)
ax1.set_ylabel(u"净值", fontproperties=font, fontsize=16)
ax1.set_title(u"因子不同十分位数分组选股净值走势", fontproperties=font, fontsize=16)

# results 转换为 DataFrame
#把字典嵌套字典转化为dataframe，外层字典keys作为columns，内层字典keys作为index
results_pd = pd.DataFrame(results).T.sort_index()

results_pd = results_pd[[ u'annualized_return', u'extra_max_drawdown',  u'turnover' ,u'extra_annualized_return']]

results_pd.columns = [u'年化收益',u'超额最大回撤',u'换手率',u'超额年化收益']
results_pd.index.name = u'十分位组别'
results_pd
```
