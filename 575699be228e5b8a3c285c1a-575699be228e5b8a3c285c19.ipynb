**Alpha #01**
$$(rank(Ts\_ArgMax(SignedPower(((returns < 0) ? stddev(returns, 20) : close), 2.), 5)) -
0.5)$$

先了解下面几个个运算符：<br>
**1. Rank(x)** ： 排列所有工具的 x 值，然后返回的数值是介于0.0和1.0之间均匀分布的浮点数。<br>
例如假设你有一个向量：<br>
A = [29, 12, 5, 7, 10]<br>
现在，将会为向量里的每个元素**按升序分级**，最小一级为 0：<br>
A = [29, 12, 5, 7, 10]<br>
A' = [4, 3, 0, 1, 2] <br>
每个“等级”再除以（元素个数 - 1）。于是得到：<br>
Rank (A) = [1.00, 0.75, 0.00, 0.25, 0.50]<br>
所以您等到了一个从 0 到 1 间距相等的向量。<br>

**2. Ts\_ArgMax(x,n):**  找出过去n天 x的值的最大值，但返回值为对应的索引<br>

**3. SignedPower(x,e):**  Sign(x) * (Abs(x)^e) ;      
Sign(x)：如果 x > 0，就返回 1，如果 x < 0，则返回 -1，如果 x == 0，返回 0<br>

**4. cond ? expr1 : expr2**  如果 cond 为 true，那么 expr1；否则 expr2。例如，close < open ? close : open<br>

根据上面的运算符理解该公式：<br>
1. (returns < 0) ? stddev(returns, 20) : close  判断每日回报率returns（这里的回报率为close-close，即收盘对收盘），如果小于0，返回前20天每日回报率的标准差；否则返回收盘价<br>
2. 对1的返回值记做x1， 求SignedPower(x1, 2.) 即保留原来的正负性质，对其值做平方处理，可以理解为将 差异放大<br>
3. 对2的返回值记做x2，rank(Ts\_ArgMax(x2,5)) 找出过去5天x2中的最大值并返回索引，再进行rank操作得到0到1的等间距序列<br>
4. 对3的返回值记做x3， x3-0.5   中性化处理，一半做多一半做空，多空力量平衡。<br>

可以看出，主要涉及的数据为每日回报率和收盘价：如果每日回报率为正，使用收盘价；为负则使用前20日每日回报率的标准差。由于平方处理使差异放大，收盘价的平方大于后者的概率极高。<br>
再仔细思量其内涵：<br>
1.由于收盘价的平方值普遍大于回报率标准差的平方值，只要过去5天有一天对日回报率为正，便以过去5天中收盘价最高的天数（索引）做权重。例如股票a在前5天中收盘价的最高值在前第5天；股票b在前第1天，那么我们给a的权重（5）就大于b的权重（1）。(平方处理后)<br>
2.而对过去5天日回报率为负的情况，将过去5天中日回报率的20日标准差最大的那一天的天数（索引）做权重。(平方处理后)<br>

我个人的理解是：<br>
首先，更大的过去20日回报率标准差意味着更高的震荡。
前5天中的高位的时间点离现在越久，且不是一路下跌，则认为会有更高的概率反弹；若高位再前第5天，且每天都下跌，则观察震荡，震荡（这里就是跌幅）最大的时间点离现在越久，则认为有更高的概率反弹。 我认为这是种 Mean Reversion 的思想：股票价格总是围绕其平均值上下波动的。 一种上涨或者下跌的趋势不管其延续的时间多长都不能永远持续下去 ，最终均值回归的规律一定会出现：涨得太多了， 就会向平均值移动下跌；跌得太多了，就会向平均值移动上升。<br>
不过这里并不是直接利用“回归到均值”这点，而是主要利用了“波动”的概念。

当然，作为一个工程界跨界来的小白，我的理解应该还不太到位；若大家有更好的想法求告知哈~
