# 正太分布的应用 VaR
代表某种资产在一定概率下的最大损失

```python
begin='20160101'
end='20180101'

data=DataAPI.MktIdxdGet(indexID=u"",ticker=u"000300",tradeDate=u"",beginDate=begin,endDate=end,exchangeCD=u"XSHE,XSHG",field=u"tradeDate,ticker,closeIndex",pandas="1")
#添加收益率序列
data['return']=data['closeIndex'].pct_change().dropna()
#去除缺失值
data=data.dropna()

data.head()
```

```python
#收益率的均值和方差
re_mean,re_var=data['return'].mean(),data['return'].var()
#计算VaR
stats.norm.ppf(0.05,re_mean,re_var**0.5)
```

说明在当日有95%的概率损失不会超过1.678%
这只是简单的计算VaR，还可以通过Arch，Garch等去估计

# t分布

```python
x=np.arange(-4,4,0.01)

#对比正太分布和自由度为5与10的T分布
plt.plot(x,stats.norm.pdf(x),label='Norm')
plt.plot(x,stats.t.pdf(x,5),label='df=5')
plt.plot(x,stats.t.pdf(x,10),label='df=10')
plt.legend(loc=0)
plt.show()
```

# F分布
