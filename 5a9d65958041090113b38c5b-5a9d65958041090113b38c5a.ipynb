思路步骤：
1.先找出相关系数大的股票对(这里选择银行板块)，相关系数大的股票才可能协整
2.判断找出来的股票对是否真有协整性
	2.1 分别检验这两个股票价格的时间序列是否平稳(ADF检验)，平稳的话ok，不平稳则进行差分直到平稳为止
    2.2 若两支股票价格的时序同阶差分后平稳(事实上股价不会平稳，不用考虑)，使用Engle-Granger两步法来检验协整性
    	2.2.1 Engle-Granger两步法(1):对这两支股票的价格做最小二乘(OLS)估计
		2.2.2 Engle-Granger两步法(2):用ADF检验回归方程的残差是否平稳，若平稳则两个时序是协整的，否则over
3.根据上一步得到的OLS方程编写策略，若残差Zt ≥ 2σ(参考论文设定)则空仓y股票全仓x股票，若残差Zt ≤ −2σ则空仓x全仓y
4.升级版策略：每天收盘后重新进行OLS估计，开平仓判断标准同上

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import seaborn as sns
from pandas import DataFrame,Series
import matplotlib.pyplot as plt
```

```python
stocks = DataAPI.EquIndustryGet(industryVersionCD=u"",industry=u"",secID=u"",ticker=u"",industryID1=u"01030321",intoDate=u"",field=u"",pandas="1")
id_set = [] #获得银行板块的所有股票，用一个list储存
for stk in stocks["secID"]:
    if stk not in id_set:
        id_set.append(stk)
```

```python
df = DataFrame()
data = DataAPI.MktEqudGet(tradeDate=u"",secID=id_set,beginDate=u"20140101",endDate=u"20151231",isOpen=u"1",pandas="1")
data.index = data.loc[:,('tradeDate')]
data = data.loc[:,('secID','closePrice')]
date = []
for index in data.index:
    if index not in date:
        date.append(index)
for stk in id_set:
    df2 = DataFrame(data[data['secID']==stk]['closePrice'])
    df2.columns = [stk]
    df = pd.concat([df,df2], axis=1)
    
df = df.dropna(axis = 1,how = 'any')
df.shape
```

在进行协整检验之前，得先对这两组时间序列分别做单位根检验，看看它们是否是同阶单整的，这样做出来协整才比较靠谱

```python
st1 = sm.tsa.stattools.adfuller(stock_df1)
if st1[0]>st1[4]['10%']:
    print('序列1不平稳')
st2 = sm.tsa.stattools.adfuller(stock_df2)
if st2[0]>st2[4]['10%']:
    print('序列2不平稳')
```

发现这两组时间序列都不平稳，我们进行差分来看看它是几阶单整的

```python
u1 = stock_df1.diff().dropna()
st_1 = sm.tsa.stattools.adfuller(u1)
if st_1[0]<st_1[4]['1%']:
    print('序列1在1%水平下一阶单整')
u2 = stock_df2.diff().dropna()
st_2 = sm.tsa.stattools.adfuller(u2)
if st_2[0]<st_2[4]['1%']:
    print('序列2在1%水平下一阶单整')
```

这两组时间序列都是一阶单整的，满足协整检验的前提

使用Engle-Granger两步协整检验法来检验协整
第一步用 OLS 法进行协整回归

```python
x = stock_df1
y = stock_df2
X = sm.add_constant(x)
result = (sm.OLS(y,X)).fit()
print(result.summary())
result.params[1]
```

```python
ut = y - 1.9577*x + 0.7354
mean = np.mean(ut)
std = np.std(ut)  

print('std = %s'%(std))

#up_line1 = Series(mean + std, index = ut.index)
horizon = Series(mean,index = ut.index)
ut.plot()
horizon.plot()
```

用ADF检验残差估计值的平稳性

```python
t = sm.tsa.stattools.adfuller(ut)

output = DataFrame(index=['Test Statistic value','p-value','Lags Used','Number of Observations Used','Critical Value(1%)','Critical Value(5%)','Critical Value(10%)'],columns=['value'])
output.iloc[0][0] = t[0]
output.iloc[1][0] = t[1]
output.iloc[2][0] = t[2]
output.iloc[3][0] = t[3]
output.iloc[4][0] = t[4]['1%']
output.iloc[5][0] = t[4]['5%']
output.iloc[6][0] = t[4]['10%']
output
```

ADF检验原假设是有单位根，这里p值小于0.01，在0.01水平下拒绝原假设：残差估计值非平稳。这两组时间序列间存在很强的协整性

由于优矿只支持获取最近30天的数据（而且每天都要重复计算确实很耗时间），所以在研究板块这里先算好再导入
这里我并没有使用未来数据，为了方便计算我把这4年两支个股的收盘价都get到，然后每天收盘根据之前两年的收盘价数据估计出回归方程，之后再进行操作

```python
close0 = DataAPI.MktEqudGet(tradeDate=u"",secID='601288.XSHG',beginDate=u"20140101",endDate=u"20171231",isOpen=u"1",pandas="1")
close1 = close0.loc[:,('closePrice')]
close0.index = close0.loc[:,('tradeDate')]
length1 = close0.iloc[:,0].size  #前两年时间长度
length2 = close0[close0['tradeDate']<'2016-01-01'].iloc[:,0].size  #总时间长度

close2 = DataAPI.MktEqudGet(tradeDate=u"",secID='601939.XSHG',beginDate=u"20140101",endDate=u"20171231",isOpen=u"1",pandas="1").loc[:,('closePrice')]

zeros = np.zeros((length1-length2,3))
data_df = DataFrame(zeros,columns = ['const','coef','std']) #用来记录每个时间窗口的ols回归的截距、系数以及残差的标准差
for i in range(0,length1-length2):
    x = close1[i:i+length2]
    y = close2[i:i+length2]
    X = sm.add_constant(x)
    result = (sm.OLS(y,X)).fit()
    ut = y - result.params[1]*x - result.params[0]
    #data_df.iloc[0][0] = 1
    data_df.iloc[i][0] = result.params[0] #第一列是截距列
    data_df.iloc[i][1] = result.params[1] #第二列是系数列
    data_df.iloc[i][2] = np.std(ut) #第三列是残差的标准差

data_df.to_csv(u'df_data.csv')
data_df
```

emmm 怎么看上去斜率系数在不断增大截距在不断减小呢，是不是有什么问题呢

看来斜率并不是一直在增大，结合之前建行和农行的走势来看，建行的波动幅度相对于农行越来越大，所以斜率系数越来越大没有问题
