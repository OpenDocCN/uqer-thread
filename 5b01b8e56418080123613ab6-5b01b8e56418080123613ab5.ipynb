作业2

```python
# 利用DataAPI.MktStockFactorsOneDayGet 取一天的某些感兴趣的因子值
factor = DataAPI.MktStockFactorsOneDayGet(
    tradeDate=u"20180518",
    field=u"secID,ticker,AccountsPayablesTDays,PE")
print('行数：%d' % len(factor))
factor.tail(1)
```

```python
# 对返回的DataFrame进行增删查改的操作

# 查看
factor.iloc[0, :2]  #按位置
factor.loc[0, ['ticker','PE']]  #按标签

# 增加
print('增加前的secID：%s' % factor.iloc[-1, 0])
factor = factor.append({'secID':'test'}, ignore_index=True)
print('增加后的secID：%s' % factor.iloc[-1, 0])

# 修改
factor.iloc[-1, 0] = 'test2'
factor.iloc[-1, 0]
print('修改后的secID：%s' % factor.iloc[-1, 0])

# 删除
factor.drop(factor.index[-1], inplace=True)  #删除行
print('删除行后的secID：%s' % factor.iloc[-1, 0])
factor.drop(factor.columns[0], axis=1, inplace=True)  #删除列
print('删除列后变为ticker：%s' % factor.iloc[-1, 0])
```

作业3

```python
# 缺失值处理

# 获取股票的申万一级行业分类
industry = DataAPI.EquIndustryGet(
    industry=u"申万",
    intoDate=u"20180518",
    field=u"ticker,industryName1")
print('行数：%d' % len(industry))
industry.head(3)
```

```python
# 合并DataFrame
factor = factor.merge(industry, on='ticker', how='left')
print('行数：%d' % len(industry))
factor.head(3)
```

```python
# 缺失值按照该股票申万一级行业的均值填充
factor.groupby('industryName1').transform(lambda x:x.fillna(x.mean()))
```
