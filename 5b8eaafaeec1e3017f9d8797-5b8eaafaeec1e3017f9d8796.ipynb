# 双动量资产配置策略的初探

<h1 style="text-align:center">导读</h1>
<p>---------------------------------------------------------------------------------------------------------------------------------</p>
<p style="font-family:arial;color:blue;font-size:15px;">A.研究的目的与背景:</p>
<p>	本篇主要集中讨论了大类资产配置的问题。资产配置(Asset Allocation)是指根据投资需求将投资资金在不同资产类别之间进行分配，通常是将资产在低风险、低收益证券与高风险、高收益证券之间进行分配。这里的证券主要是指股票（包括发达国家的股票与新兴国家的股票，债券，黄金，商品期货，外汇，房地产信托等），相比于传统的投资方式，这些证券之间的相关性较低，能起到消除非系统风险的作用。
	动量效应是指，投资者可以通过买入过去收益率高的证券、卖出过去收益率低的证券获利，这种利用证券动量效应构造的投资策略称为动量投资策略。动量效应在股票市场上存在的历史很长，并且普遍存在于世界各地的股票市场上，甚至一些近期的研究发现动量效应也存在于其它类型的交易市场上。
	本文采用一种双动量的策略进行大类资产配置的择时。双动量指的是绝对动量与相对动量。通过对于双动量的比较，选择相应的资产。本文选取中企债指数（000013.ZICN）,国债指数（000012.ZICN）,澳大利亚标普500（AXJO.XASX）,标普500（SPX.ZIUS）,日经225（N225.ZIJP）,英国富时（FTSE.XLON）,法国CAC（FCHI.ZIFR）,德国DAX（GDAXI.ZIDE）,印度（SENSEX.ZIIN）,沪深300（000300.ZICN）,香港恒生指数（HSI.ZIHK）作为测试标的。</p>
<p style="font-family:arial;color:blue;font-size:15px;">B.研究结论:</p>
<p>	1.双动量策略能在基准策略的基础上改进风险水平（更低的波动率，更高的夏普比率，以及calmar比率），同时取得相对于债券指数更高的收益。
	2.双动量策略对于三个参数：选股个数影响性能不大，最佳是3支；超额收益计算日期，对于性能影响最大，其中以60天最佳；资金占比影响风险指标，根据客户风险喜好确定。
	3.双动量策略最大取得1.68的夏普比率（年化收益11.17%,波动率4.44%,最大回撤6.41%,calmar比率为1.74）的效果，最大calmar比率为2.41(年化收益10.77%,波动率4.44%，最大回撤4.47%，夏普比率1.59）。</p>

<h1 style="text-align:center;color:blue">4.计算标的性能</h1>
<p>---------------------------------------------------------------------------------------------------------------------------------</p>
<p>	下面一段代码计算出各个标的的年化收益，波动率，最大回撤，夏普比率和Calmar比率。有表格可知，债券的年化收益偏低（3.66%与6.17%）,但是波动率较低（1.64%，3.24%），且最大回撤较小（1.19%，2.48%），风险较低。而股票指数波动率大多在20%以上，回撤达到60%，单相对的收益率较高。如果选取适当的择时策略，通过选择强势的标的，卖出弱势的标的，那将会平衡风险和收益，改善性能指标。</p>

```python
import pandas as pd
import numpy as np
from CAL.PyCAL import *
cal=Calendar('China.SSE')
path='./ms/'
def max_down(return_idx):
    tradeDate_list = list(return_idx['tradeDate'].drop_duplicates())
    max_ = -10000
    max_down=0
    return_idx_use = return_idx.set_index('tradeDate')
    for tradeDate in tradeDate_list:
        if(max_<return_idx_use['closeIndex'][tradeDate]):
            max_ = return_idx_use['closeIndex'][tradeDate]
        max_down_tmp =1.0 -  1.0*return_idx_use['closeIndex'][tradeDate]/max_
        if(max_down_tmp>max_down):
            max_down = max_down_tmp
    return max_down
    
def return_dev(return_idx,tradeDate='2008-01-01'):
    return_idx_data = return_idx[return_idx['tradeDate']>tradeDate]
    idx_name_set = list(return_idx_data['indexID'].drop_duplicates())
    idx_return = pd.DataFrame(index=[(idx_name_set)],columns= ['return','risk','sharp','max_down','calmar'])
    for idx_name in idx_name_set:
        return_idx_one  = return_idx_data[return_idx_data['indexID']==idx_name]
        idx_return['return'][idx_name] = np.mean((return_idx_one['closeIndex']/return_idx_one['closeIndex'].shift(240)-1).dropna())
        idx_return['risk'][idx_name] = np.std((return_idx_one['closeIndex']/return_idx_one['closeIndex'].shift(240)-1).dropna())
        idx_return['sharp'][idx_name] = (idx_return['return'][idx_name]-idx_return['return']['000012.ZICN'])/idx_return['risk'][idx_name]
        idx_return['max_down'][idx_name] = max_down(return_idx_one)
        idx_return['calmar'][idx_name] = (idx_return['return'][idx_name])/idx_return['max_down'][idx_name]
    return idx_return

idx = pd.read_csv(path+'idx_data.csv')
return_dev(idx)
```

<h1 style="text-align:center;color:blue">5.测试双动量策略的性能</h1>
<p>---------------------------------------------------------------------------------------------------------------------------------</p>
<p>	这里给出双动量策略，基本框架如下：</p>
<p>	1、回测时间段为2008/09/30 ~ 2017/9/30</p>
<p>	2、样本空间为中企债指数（000013.ZICN）,澳大利亚标普500（AXJO.XASX）,标普500（SPX.ZIUS）,日经225（N225.ZIJP）,英国富时（FTSE.XLON）,法国CAC（FCHI.ZIFR）,德国DAX（GDAXI.ZIDE）,印度（SENSEX.ZIIN）,沪深300（000300.ZICN）,香港恒生指数（HSI.ZIHK）指数标的。</p>
<p>	3、每次买入K（全部为1.0）资金量的股指，剩下的资金买入债券指数。</p>
<p>	4、组合按月调仓</p>
<p>		4.1、首先剔除N天收益率大于N天债券收益的股指（绝对动量）。</p>
<p>		4.2、将剔除下来的股指选择N天收益率最大的M支股票买入（相对动量），每支指数为K/M的资金。如果剔除后的股票不满M支，则全部买入，每支指数为K/M的资金。</p>
<p>	5、监控买入的股票一旦N天收益率小于N天债券收益，则卖出并买入债券。</p>


<p>	下面的函数excess_return_init(return_idx,delay_day,path,is_save)是计算delay_day日超额收益的函数</p>

```python
import pandas as pd
import numpy as np
from datetime import datetime
from datetime import timedelta
from CAL.PyCAL import *
cal=Calendar('China.SSE')
path='./ms/'

def excess_return_init(return_idx,delay_day,path='./ms/',is_save=False):
    idx_name_set = list(return_idx['indexID'].drop_duplicates())
    return_ = return_idx[return_idx['indexID']==return_idx['indexID'][0]]
    return_ =  return_[['tradeDate']]
    return_ =return_.set_index('tradeDate')
    for idx_name in idx_name_set:
        tradeDate_list = list(return_idx[return_idx['indexID']==idx_name].set_index('tradeDate').index)
        return_shift_use = return_idx[return_idx['indexID'] == idx_name]
        return_shift_use = return_shift_use.set_index('tradeDate')
        return_cell_=pd.DataFrame(index=return_shift_use.index)
        return_cell_[idx_name]=np.nan
        for i in range(len(return_shift_use)-delay_day):
            tradeDate = tradeDate_list[i+delay_day]
            lastDate = tradeDate_list[i]
            tradeDate_index = pd.Index([tradeDate])
            lastDate_index = pd.Index([lastDate])            
            return_cell_[idx_name][tradeDate] = return_shift_use['closeIndex'][tradeDate_index].values[0]/return_shift_use['closeIndex'][lastDate_index].values[0]-1.0
        return_ = return_.join(return_cell_,how='outer')
    return_ = return_.tail(len(return_)-delay_day)
    return_ = return_.fillna(method='ffill')
    if(is_save):
        return_.to_csv(path+'excess_return_'+str(delay_day)+'.csv')
    return return_
```

<p>	下面是回测代码，函数portfolio_back_test与portfolio_back_test_for_base是回测的主函数，前者是双动量的回测函数，后者是基准策略回测函数,基准策略是用K的资金买入所有的股票指数，用剩余的资金买入债券指数，并一直持有不换仓。</p>

```python
import pandas as pd
import numpy as np
from CAL.PyCAL import *
cal=Calendar('China.SSE')
path='./ms/'

def excess_return_daliy(return_use,index_num,theld):
    ret_sort=pd.DataFrame()
    ret_sort['ret'] = return_use.values[0][1:]
    ret_sort['indexID'] = return_use.columns.values[1:]
    ret_sort=ret_sort[ret_sort['ret']>theld]
    ret_sort_h = ret_sort
    ret_sort = ret_sort.sort_values(by=['ret']).tail(index_num)
    return ret_sort


def portfolio_back_test(return_idx,return_,delay,begin_date,end_date,debet ='000013.ZICN',index_num=3,N=20,theld=0.03,portfolio_proportion=0.7,Index_daily = excess_return_daliy,path='./ms/',is_Save=False):

# 每天要修改的表：
#     portfolio_sec:每天持有的指数
#     portfolio_lasttrade_cell:上一天持仓指数的比重
#     portfolio_cell:本日持有的证券
#     portfolio_:每一天的持仓比重
   
    portfolio_sec = dict()
    idx_name_set = list(return_idx['indexID'].drop_duplicates())
    tradeDate_list = list(idx[idx['indexID']==debet].set_index('tradeDate').index)
    portfolio_ = pd.DataFrame(columns=idx_name_set)
    if(begin_date not in tradeDate_list or end_date not in tradeDate_list):
        print('the date is not right')
        return portfolio_
    else:
        i_begin = tradeDate_list.index(begin_date)
        i_end = tradeDate_list.index(end_date)
    i = i_begin
    tradeDate = tradeDate_list[i_begin]
    return_idx_test = return_idx[return_idx['indexID']==debet]
    tradeDate_count=0
    portfolio_sec[tradeDate_list[i_begin-1]]=[]
    portfolio_lasttrade_cell = pd.DataFrame(index=[tradeDate],columns=idx_name_set)
    debet_daliy = return_idx[return_idx['indexID']==debet].set_index('tradeDate')
    while(tradeDate<end_date):
        tradeDate = tradeDate_list[i]
        last_date = tradeDate_list[i-1]
        delay_date  = tradeDate_list[i-delay]
        #if(tradeDate=='2014-07-30'):
            #oo=1
        portfolio_cell = pd.DataFrame(index=[tradeDate],columns=idx_name_set)
        return_use = return_[return_['tradeDate']==tradeDate]
        tradeDate_index = pd.Index([tradeDate])
        last_date_index = pd.Index([last_date])
        delay_date_index = pd.Index([delay_date])
        delay
        return_idx_daily = return_idx[return_idx['tradeDate']==tradeDate].set_index('indexID')
        if(len(return_idx_daily)<len(idx_name_set) and tradeDate_count==0):
            tradeDate_count=N+1
        del return_use['000012.ZICN']
        del return_use['000013.ZICN']     
        theld = (debet_daliy['closeIndex'][tradeDate_index].values[0]/debet_daliy['closeIndex'][delay_date_index].values[0] - 1.0)
        ret_sort = excess_return_daliy(return_use,index_num,theld)
        ret_sort_h = ret_sort
        ret_sort = ret_sort.sort_values(by=['ret']).tail(index_num)
        stock_index_num = len(ret_sort)
        if(tradeDate_count==0):#月末换仓
            if(len(portfolio_sec[last_date])==0): #上一日未有持仓的证券
                portfolio_cell[debet][tradeDate_index] = (1.0-stock_index_num*portfolio_proportion/index_num)*return_idx_daily['closeIndex'][debet]/return_idx_daily['openIndex'][debet]
                for indexID in ret_sort['indexID']:
                    portfolio_cell[indexID][tradeDate_index] = portfolio_proportion/index_num*return_idx_daily['closeIndex'][indexID]/return_idx_daily['openIndex'][indexID] 
                portfolio_cell =portfolio_cell.fillna(0)
                set_tmp=list(ret_sort['indexID'].values)
                set_tmp[0:0] =[debet]
                portfolio_sec[tradeDate] = set_tmp
                portfolio_lasttrade_cell = portfolio_cell
                portfolio_lasttrade_cell.index.values[0] = tradeDate
            else: #上一日有持仓的证券
                cash = 0
                cash =  cash +portfolio_.ix[last_date_index][debet]*return_idx_daily['openIndex'][indexID]/return_idx_daily['preCloseIndex'][indexID]
                for indexID in portfolio_sec[last_date]:
                    if(indexID!=debet):
                        cash = cash + portfolio_.ix[last_date_index][indexID]*return_idx_daily['openIndex'][indexID]/return_idx_daily['preCloseIndex'][indexID]
                portfolio_cell[debet][tradeDate_index] = (cash-stock_index_num*portfolio_proportion*cash/index_num)*return_idx_daily['closeIndex'][debet]/return_idx_daily['openIndex'][debet]
                for indexID in ret_sort['indexID']:
                    portfolio_cell[indexID][tradeDate_index] = cash*portfolio_proportion/index_num*return_idx_daily['closeIndex'][indexID]/return_idx_daily['openIndex'][indexID]
                set_tmp=list(ret_sort['indexID'].values)
                set_tmp[0:0] =[debet]
                portfolio_sec[tradeDate] = set_tmp
                portfolio_cell =portfolio_cell.fillna(0)
                portfolio_lasttrade_cell = portfolio_cell
                portfolio_lasttrade_cell.index.values[0] = tradeDate
        else: #非月末换仓
            flag=0
            not_in_list=[]
            in_list=[]
            for indexID in portfolio_sec[last_date]:
                if(indexID not in list(ret_sort_h['indexID']) and indexID != debet):
                    not_in_list[0:0] = [indexID]
                    flag=1
                else:
                    in_list.append(indexID)
            if(flag==1):
                for indexID in in_list:
                    if(indexID in return_idx_daily.index):
                        portfolio_cell[indexID][tradeDate_index] = portfolio_lasttrade_cell[indexID][last_date_index]*return_idx_daily['closeIndex'][indexID]/return_idx_daily['preCloseIndex'][indexID]
                    else:
                        portfolio_cell[indexID][tradeDate_index] = portfolio_lasttrade_cell[indexID][last_date_index]
                cash=0
                if(debet in return_idx_daily.index):
                    cash =  cash +portfolio_.ix[last_date_index][debet]*return_idx_daily['openIndex'][debet]/return_idx_daily['preCloseIndex'][debet]
                else:
                    cash = portfolio_.ix[last_date_index][debet]
                for indexID in not_in_list:
                    if(indexID in return_idx_daily.index):
                        cash = cash + portfolio_lasttrade_cell[indexID][last_date_index]*return_idx_daily['openIndex'][indexID]/return_idx_daily['preCloseIndex'][indexID]
                    else:
                        cash = cash + portfolio_lasttrade_cell[indexID][last_date_index]
                portfolio_cell[debet][tradeDate_index] = cash*return_idx_daily['closeIndex'][debet]/return_idx_daily['openIndex'][debet]
                portfolio_sec[tradeDate] = in_list
                portfolio_cell = portfolio_cell.fillna(0)
            else:
                portfolio_cell[debet][tradeDate_index] = portfolio_lasttrade_cell[debet][last_date_index]*return_idx_daily['closeIndex'][debet]/return_idx_daily['preCloseIndex'][debet]
                for indexID in portfolio_sec[last_date]:
                    if(indexID in return_idx_daily.index):
                        portfolio_cell[indexID][tradeDate_index] = portfolio_lasttrade_cell[indexID][last_date_index]*return_idx_daily['closeIndex'][indexID]/return_idx_daily['preCloseIndex'][indexID]
                    else:
                        portfolio_cell[indexID][tradeDate_index] = portfolio_lasttrade_cell[indexID][last_date_index]
                portfolio_cell = portfolio_cell.fillna(0)
                portfolio_sec[tradeDate] = portfolio_sec[last_date]
            portfolio_lasttrade_cell = portfolio_cell
            portfolio_lasttrade_cell.index.values[0] = tradeDate

        portfolio_ = pd.concat([portfolio_,portfolio_cell],axis=0)
        if(tradeDate_count>N):
            tradeDate_count = 0
        else:
            tradeDate_count = tradeDate_count +1            
        i = i+1
    if(is_Save):
        portfolio_.to_excel(path+'results_'+str(delay)+'_'+str(index_num)+'_'+str(portfolio_proportion)+'.xlsx')
    return portfolio_

def portfolio_back_test_for_base(return_idx,begin_date,end_date,debet =['000013.ZICN','000012.ZICN'],choice=0,N=20,portfolio_proportion=0.7,path='./ms/',is_Save=False):

# 每天要修改的表：
#     portfolio_sec:每天持有的指数
#     portfolio_lasttrade_cell:上一天持仓指数的比重
#     portfolio_cell:本日持有的证券
#     portfolio_:每一天的持仓比重

    portfolio_sec = dict()
    idx_name_set = list(return_idx['indexID'].drop_duplicates())
    tradeDate_list = list(idx[idx['indexID']==debet[choice]].set_index('tradeDate').index)
    portfolio_ = pd.DataFrame(columns=idx_name_set)
    if(begin_date not in tradeDate_list or end_date not in tradeDate_list):
        print('the date is not right')
        return portfolio_
    else:
        i_begin = tradeDate_list.index(begin_date)
        i_end = tradeDate_list.index(end_date)
    i = i_begin
    tradeDate = tradeDate_list[i_begin]
    portfolio_sec[tradeDate_list[i_begin-1]]=[]
    portfolio_lasttrade_cell = pd.DataFrame(index=[tradeDate],columns=idx_name_set)
    while(tradeDate<end_date):
        tradeDate = tradeDate_list[i]
        last_date = tradeDate_list[i-1]
        portfolio_cell = pd.DataFrame(index=[tradeDate],columns=idx_name_set)
        tradeDate_index = pd.Index([tradeDate])
        last_date_index = pd.Index([last_date])
        return_idx_daily = return_idx[return_idx['tradeDate']==tradeDate].set_index('indexID')   
        if(len(portfolio_sec[last_date])==0): #上一日未有持仓的证券
            portfolio_cell[debet[choice]][tradeDate_index] = (1.0-portfolio_proportion)*return_idx_daily['closeIndex'][debet[choice]]/return_idx_daily['openIndex'][debet[choice]]
            for indexID in idx_name_set:
                if(idx_name_set not in debet):
                    portfolio_cell[indexID][tradeDate_index] = portfolio_proportion/(len(idx_name_set)-len(debet))*return_idx_daily['closeIndex'][indexID]/\
                    return_idx_daily['openIndex'][indexID] 
            portfolio_cell =portfolio_cell.fillna(0)
            set_tmp = idx_name_set
            for debet_ in debet:
                set_tmp.remove(debet_)
            set_tmp.append(debet[choice])
            portfolio_sec[tradeDate] = set_tmp
            portfolio_lasttrade_cell = portfolio_cell
            portfolio_lasttrade_cell.index.values[0] = tradeDate
        else: #上一日有持仓的证券
            portfolio_cell[debet[choice]][tradeDate_index] = portfolio_lasttrade_cell[debet[choice]][last_date_index]*return_idx_daily['closeIndex'][debet[choice]]/\
            return_idx_daily['preCloseIndex'][debet[choice]]
            for indexID in portfolio_sec[last_date]:
                if(indexID in return_idx_daily.index):
                    portfolio_cell[indexID][tradeDate_index] = portfolio_lasttrade_cell[indexID][last_date_index]*return_idx_daily['closeIndex'][indexID]/\
                    return_idx_daily['preCloseIndex'][indexID]
                else:
                    portfolio_cell[indexID][tradeDate_index] = portfolio_lasttrade_cell[indexID][last_date_index]
            portfolio_cell = portfolio_cell.fillna(0)
            portfolio_sec[tradeDate] = portfolio_sec[last_date]
            portfolio_lasttrade_cell = portfolio_cell
            portfolio_lasttrade_cell.index.values[0] = tradeDate

        portfolio_ = pd.concat([portfolio_,portfolio_cell],axis=0)          
        i = i+1
    if(is_Save):
        portfolio_.to_excel(path+'results_'+str(portfolio_proportion)+'_base'+'.xlsx')
    return portfolio_
```

```python
import pandas as pd
import numpy as np
path='./ms/'
def cal_prefmore(return_idx,name='ljy',period_long=10.5):
    pref = pd.DataFrame(index=[name],columns=['return','risk','max_down','sharpe','calmar'])
    pref['return'] = pow(return_idx[-1]/return_idx[0],1.0/period_long)-1
    pref['risk'] = np.std(return_idx[120:]/return_idx[0:-120]-1)
    max_=-0.0
    max_down = 0.0
    for i in range(len(return_idx)):
        if(max_<return_idx[i]):
            max_ = return_idx[i]
        if(1.0-1.0*return_idx[i]/max_>max_down):
            max_down = 1.0-1.0*return_idx[i]/max_
    pref['max_down'] = max_down
    pref['sharpe'] = (pref['return']-0.037)/pref['risk']
    pref['calmar'] = (pref['return'])/pref['max_down']
    return pref

account = pd.read_excel(path+'results2.xlsx')
perf_total = pd.DataFrame()
for test_name in account.columns:
    return_idx = np.array(account[test_name])
    perf = cal_prefmore(return_idx,test_name)
    perf_total = pd.concat([perf_total,perf],axis=0)
perf_total.to_excel(path+'results3.xlsx')
perf_total
```
