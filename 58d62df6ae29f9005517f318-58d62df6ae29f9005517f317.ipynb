```python
import matplotlib.pyplot as plt
import seaborn
import pandas as pd
# from sqlalchemy import create_engine
```

```python
# uqer.datayes.com的基金分类：E-股票型，H-混合型，B-债券型，SB-短期理财债券，M-货币型

# subplots的使用，返回一个figure和ax对象，其中此时的ax是[[ax0,ax1],[ax2,ax3]]
fig, ax = plt.subplots(2, 2, sharex=True,sharey=True)

# 动态调整fig的大小
fig.set_size_inches(18,12)

# ax的解释
((ax0, ax1), (ax2, ax3)) = ax
category_list = ['E','H','B','M']
for num,each_type in enumerate(category_list): # enumerate使得返回序号和内容
    num_x = int(num / 2) #取整
    num_y = num % 2 #取余
    df = DataAPI.FundGet(etfLof=u"",secID=u"",ticker=u"",category=each_type,idxID=u"",idxTicker=u"",listStatusCd=u"",operationMode=u"",field=u"",pandas="1")
    df['establishDate'] = pd.to_datetime(df['establishDate'])
    df = df.set_index('establishDate')
    x = pd.DatetimeIndex.to_pydatetime(df.resample('as').count().index) # 一种比较简便的转换datetime64[ns]至python的datetime。因为在matplotlib中python自己的datetime比numpy的datetime64[ns]更智能。
    y = df.resample('as').count()['secID'].values
    ax[num_x][num_y].bar(x,y,width=10)
    ax[num_x][num_y].set_title("Fund_"+each_type)
plt.show()
```
