脉冲比=当日成交额 / MA(过去 5 个交易日的成交额)
成交量出现正的脉冲，当天股市行情上涨的概率较大。总体趋势是脉冲比越大，当天上涨概率越高，总体概率在 76.1%。
成交量的脉冲出现时，往往是有力量大量买入股票造成的，这往往隐藏着A股未来向好的征兆。
我们可以根据脉冲比，设置参数：1.脉冲比阈值（例如：1.3）。2.持有期限（例如：30天）。如果持有期内，出现新的大于阈值的脉冲，则重现计算持有期限。
通过验证我们可以看出，基于成交量脉冲的大盘择时策略在净值收益上存在一定的优势，我们可以根据脉冲比的大小来判断之后一段时间股票价格的走向，进而取得更好的收益。

```python
para={'times': 1.3,# 成交额是MA(过去5个交易日的成交额，不含当日)的倍数
          "N":5,#N日均额,'N-Day'average turnoverValue
              }
```

```python
import pandas as pd
def get_vtimes(ticker="002004",**para):
    df=DataAPI.MktEqudAdjGet(ticker=ticker, beginDate="20221001",endDate="20230104",tradeDate=u"",\
                                 isOpen="1",field=u"ticker,secShortName,tradeDate,openPrice,closePrice,highestPrice,lowestPrice,\                           preClosePrice,turnoverVol,turnoverValue",pandas="1")
    
    #print df
    df.rename(columns={'highestPrice':'H','openPrice':'O','lowestPrice':'L','closePrice':'C','turnoverValue':'amount'},inplace=True)
    if len(df)==0 :
        return pd.DataFrame()
    for i in [1,3,5,7,9,15,30]:
        ma='Yield_in_the_next_{}days'.format(i)
        df[ma]=(df["C"].shift(-i)-df["C"])/df["C"]
    maamount='ma_Value{}'.format(para["N"])
    df[maamount]=df["amount"].shift(1).rolling(para["N"]).mean()
    #成交额大于过去N=5日均额df[maamount]的倍数times
    df["times"]=1.0*df["amount"] /df[maamount]
    #当天上涨
    df["return"]=(df["C"]-df["C"].shift(1))/df["C"].shift(1)
    con = df["amount"] >para["times"]*df[maamount]
    df.loc[con,"Vtimes"]=1
    return df
```
