```python
import pandas as pd
from datetime import *
import time
from CAL.PyCAL import *

BIG_IS_GOOD=False
SMALL_IS_GOOD=True
universe =  DynamicUniverse('HS300')
factor_list = ["Aroon", "CR20", "PVT6", "ILLIQUIDITY"]
factor_config = {'Aroon':BIG_IS_GOOD,'CR20':BIG_IS_GOOD,'PVT6':BIG_IS_GOOD,'ILLIQUIDITY':SMALL_IS_GOOD}  # True表示顺序(越小越好)，False表示逆序(越大越好)
factor_filename = r'./factor_300.pkl'
```

# 1. 为了性能，从api提取数据，保存到pickle中
pickle比起csv来文件尺寸较小，uqer给的存储空间只有1个G, 要省着用。
我们保存的文件并不需要人眼去看， 没必要为了可读去牺牲空间
* 这一步只需要做一次。
* 只有因子列表改变， 或者想变更时间段时才需要重新做。

```python
def save_factors(start, end, universe, factor_list):
    
    cal = Calendar('China.SSE')
    span = Period('1D')
    date1 = datetime.strptime(start, '%Y-%m-%d')
    date2 = datetime.strptime(end, '%Y-%m-%d')

    # 生成universes,包含DynamicUniverse历史上任何时点的成分股
    # 所以真正的universe要比当前的成分股数量大的多， 一般在2倍以上， 因此空间和内存紧张的话，要慎重选择比较大的板块
    univ_set = set()
    while date1<= date2:
        date1str = date1 if isinstance(date1, str) else date1.strftime('%Y-%m-%d')
        univ = universe.preview(date1)
        univ_set.update(univ)
        date1 = cal.advanceDate(date1,span).toDateTime()

    univ_list = list(univ_set)

    df_allday_alltick_factors = pd.DataFrame()
    for i in range(len(univ_list)//50+1):
        print (i*50, (i+1)*50)
        univ_list_piece = ','.join(univ_list[i*50: (i+1)*50])
        fields = u"tradeDate,secID," + ','.join(factor_list)
        for _ in range(10):
            try:
                p=DataAPI.MktStockFactorsDateRangeGet(secID=univ_list_piece,ticker=u"",beginDate=start,endDate=end
                                                      ,field=fields
                                                      ,pandas="1")
                break
            except Exception as e:
                if _ < 9:
                    time.sleep(1)
                    continue
                else:
                    raise(e)
        df_allday_alltick_factors = pd.concat([df_allday_alltick_factors, p], join='outer')
    df_allday_alltick_factors.to_pickle(factor_filename)
        
start = '2007-01-01'  
end = datetime.today().strftime('%Y-%m-%d')

save_factors(start, end, universe, factor_list)
```

# 2. 从pickle文件读取因子， 环境重启后需要执行， 但回测时不必重复读取。

```python
import pandas as pd
df_allday_alltick_factors = None

def init_factors():
    global df_allday_alltick_factors
    df_allday_alltick_factors= pd.read_pickle(factor_filename)
    df_allday_alltick_factors.set_index(['tradeDate', 'secID'], inplace=True)

init_factors()
```
