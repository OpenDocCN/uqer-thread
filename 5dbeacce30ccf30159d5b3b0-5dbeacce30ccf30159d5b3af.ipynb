# 概率角度的投资-频繁模式

## 频繁模式

```python
import numpy as np
import pandas as pd
import datetime
import time
import matplotlib.pyplot as plt
```

```python
## 股票选取 
## 一定要自己用复权数据计算，优矿原始数据的涨跌幅经常出现奇奇怪怪的数字
def chs_factor(ticker = '000010',begin = '20010101',end = '20170201', field = ['tradeDate' ,'closePrice', 'preClosePrice']):
    dataday = DataAPI.MktEqudAdjGet(tradeDate=u"",secID=u"",ticker=ticker,beginDate= begin,endDate= end,isOpen="1",field= field,pandas="1")
    dataday['chgPct'] = dataday['closePrice'] / dataday['preClosePrice'] - 1
    return dataday
```

```python
df_ori = chs_factor()
df_ori['chgPct'].hist()
```

上一篇，讨论了条件概率在股票择时中的应用。做的比较粗糙。这次，将按照更加精细的划分改进条件概率方法在择时中的使用。

上一次说到，投资决策应该是，$p(\text{up} | info) > p( \text{down} | info)$
但是，这种投资还不够细腻。上涨一个点和上涨5个点绝不是一种概念，对此，我们更加精细的对$p( \text{up} | info)$进行划分。比如，涨的多与涨的少应该要进行区分。

```python
## 先利用分位数的方法的到其标识点
stat = pd.qcut(df_ori['chgPct'], 3, labels = [-1, 0, 1])
delay_day = 4

# 统计模式
stat_dict = {}
for i in range(stat.shape[0] - delay_day):
    st = tuple(stat.values[i + j] for j in range(delay_day))
    if st not in stat_dict.keys():
        stat_dict[st] = 1
    stat_dict[st] += 1
    
# 统计前模式
stat_pre_dict = {}
for i in range(stat.shape[0] - delay_day + 1):
    st = tuple(stat.values[i + j] for j in range(delay_day - 1))
    if st not in stat_pre_dict.keys():
        stat_pre_dict[st] = 1
    stat_pre_dict[st] += 1

# 统计概率
result_dict = {}
for pre_key in stat_pre_dict.keys():
    stat_sum = np.float(stat_pre_dict[pre_key])
    result_dict[pre_key] = {}
    for key in stat_dict.keys():
        if pre_key == key[:-1]:
            result_dict[pre_key][key[-1]] = stat_dict[key] / stat_sum
```

```python
up_index = [key for key in result_dict.keys() if result_dict[key].get(1, 0) == max(result_dict[key].values())]
sigals = []
for i in range(stat.shape[0] - (delay_day - 1)):
    st = tuple(stat.values[i + j] for j in range(delay_day - 1))
    if st in up_index:
        sigals.append(1)
    else:
        sigals.append(0)
df_ori['signal'] = [0]*(delay_day - 1) +  sigals
```

```python
## 训练集上的表现
plt.figure(figsize= (10, 5))
(df_ori['chgPct']+ 1).cumprod().plot(label = 'ori_ser')
(df_ori['chgPct'] * df_ori['signal'] + 1).cumprod().plot(label = 'stratage')
plt.legend()
```

从上面的分布，可以看到，有些模式的出现频率极少，我们认为这种模式是偶尔出现的，并不具别频繁性。由此，我们引出概念：**频繁模式**

频繁模式是指，在数据中经常性出现的模式，只有经常出现，我们才能将其认为是一种可信的模式。

```python
## 先利用分位数的方法的到其标识点
stat = pd.qcut(df_ori['chgPct'], 5, labels = [-2, -1, 0, 1, 2])
delay_day = 4

# 统计模式
stat_dict = {}
for i in range(stat.shape[0] - delay_day):
    st = tuple(stat.values[i + j] for j in range(delay_day))
    if st not in stat_dict.keys():
        stat_dict[st] = 1
    stat_dict[st] += 1
    
# 统计前模式
stat_pre_dict = {}
for i in range(stat.shape[0] - delay_day + 1):
    st = tuple(stat.values[i + j] for j in range(delay_day - 1))
    if st not in stat_pre_dict.keys():
        stat_pre_dict[st] = 1
    stat_pre_dict[st] += 1
    
## 排除非频繁模式
thread_num = np.percentile(stat_pre_dict.values(), 0.1)
print(thread_num)
for key in stat_pre_dict.keys():
    if stat_pre_dict[key] < thread_num:
        stat_pre_dict.pop(key)


# 统计概率
result_dict = {}
for pre_key in stat_pre_dict.keys():
    stat_sum = np.float(stat_pre_dict[pre_key])
    result_dict[pre_key] = {}
    for key in stat_dict.keys():
        if pre_key == key[:-1]:
            result_dict[pre_key][key[-1]] = stat_dict[key] / stat_sum
```

```python
up_index = [key for key in result_dict.keys() if result_dict[key].get(2, 0) == max(result_dict[key].values())]
sigals = []
for i in range(stat.shape[0] - (delay_day - 1)):
    st = tuple(stat.values[i + j] for j in range(delay_day - 1))
    if st in up_index:
        sigals.append(1)
    else:
        sigals.append(0)
df_ori['signal'] = [0]*(delay_day - 1) +  sigals
```

```python
## 训练集上的表现
plt.figure(figsize= (10, 5))
(df_ori['chgPct']+ 1).cumprod().plot(label = 'ori_ser')
(df_ori['chgPct'] * df_ori['signal'] + 1).cumprod().plot(label = 'stratage')
plt.legend()
```
