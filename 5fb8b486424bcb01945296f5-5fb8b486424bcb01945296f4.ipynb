```python
def calcUnit(portfolio_value,ATR):
    '''
     计算unit,注意股数为100的整数倍
    '''
    value = portfolio_value * trade_percent         
    return int((value*0.01/ATR)/coef)                            
```

```python
def timing_turtle(context):
    #全局变量声明
    global record                                                                                 #声明record为局变量
   
    #数据获取（通用部分）：投资者账户，可供投资股票，价格数据，持仓数据，账户金额数据
    account = context.get_account('fantasy_account')                                              #获取投资者的股票账户（fantasy_account）
    current_futures = context.get_universe(asset_type = 'futures')               #获取当前所有主力期货合约列表（universe）
    current_universe=[]
    for symbol in current_futures:
        symbol = context.get_symbol(symbol)
        current_universe.append(symbol)
        
        if context.mapping_changed(symbol):
            
            
            symbol_before, symbol_after = context.get_rolling_tuple(symbol)
            if  account.get_position(symbol_before):
                account.switch_position(symbol_before, symbol_after)
                
                add_unit = float(record[record['symbol'] == symbol_before]['add_time'])
                last_price =float(record[record['symbol'] == symbol_before]['last_buy_price'])
                record.append(pd.DataFrame({'symbol':[symbol_after],'add_time':[add_unit],'last_buy_price':[last_price]}))
                record = record[record['symbol']!=symbol_before]
        
            
            
    history = context.history(current_universe,['closePrice','lowPrice','highPrice'],Max_time_range, freq='30m', rtype='array')   #构建海龟交易系统时考虑过去Max_time_range的历史数据
    security_position = account.get_positions()                                                   #字典数据，上一K线结束后的有效证券头寸，即持仓数量大于0的证券及其头寸
    #print(security_position)
    cash = account.cash                                                                           #获取股票账户可用于投资的现金额度
    #按照海龟交易系统循环处理所有证券池中股票
    for sec in current_universe:                                                                  #遍历所有可供投资股票；注：如该策略执行了选股策略，该部分应遍历被选中股票     
    #数据获取（专有部分）：最低价，最高价，收盘价
        
        close = np.array(history[sec]['closePrice'],dtype=float)                                                          #获取股票sec过去Max_time_range天的收盘价
        low = np.array(history[sec]['lowPrice'],dtype=float)                                                           #获取股票sec过去Max_time_range天的最低价
        high = np.array(history[sec]['highPrice'],dtype=float)                                                           #获取股票sec过去Max_time_range天的最高价
        current_price = context.current_price(sec)                                                  #获得当前时刻价格
        # print(context.current_price(sec))
    #数据处理（策略部分）：计算真实波幅，
        atr = ta.ATR(high, low, close, atrlength)[-1]                                               #计算真实波幅 
        ma10=ta.MA(close,10)
    #策略部分(入场): 突破DC通道上轨，入场，买入1个单位的期货
        if  current_price> high[-DC_range:-1].max() and sec not in security_position:                                  # 收盘价上穿DC上轨，且无持仓（突破通道上轨阻力位）下面的elif 表示有持仓时的逻辑
            unit = calcUnit(account.portfolio_value,atr)                                                               # 计算建仓时应购入的期货的数量
            account.order(sec,unit,'open')                                                                                         # 买入unit手的sec期货
            if  len(record)!=0:                                                                 
                record = record[record['symbol']!=sec]                                                                  # 清空record中sec过期的记录                             
            record = record.append(pd.DataFrame({'symbol':[sec],'add_time':[1],'last_buy_price':[current_price]}))     # 记录股票，加仓次数及买入价格
            continue  #回到for 语句
    #策略部分(加仓)：若股价在上一次买入（或加仓）的基础上上涨了0.5N，则加仓1个单位的股票
        elif    sec in security_position:                                                          # 判断是否已经持仓
                # print(security_position)
                # print(context.now)
                print((record[record['symbol'] == sec]))
                last_price =float(record[record['symbol'] == sec]['last_buy_price'])             # 上一次的买入价格
                
                add_price =last_price + 0.5 * atr                                                # 计算是否加仓的判断价格
                add_unit = float(record[record['symbol'] == sec]['add_time'])                         # 已加仓次数
                if  current_price > add_price and add_unit < limit_unit:                                # 价格上涨超过0.5N并且加仓次数小于4次
                    unit = calcUnit(account.portfolio_value,atr)                                        # 计算加仓时应购入的股票数量
                    account.order(sec,unit,'open')
                    # print(account.get_orders())
                    # 买入1unit的股票
                    record.loc[record['symbol']== sec,'add_time']=record[record['symbol']== sec]['add_time']+1   # 加仓次数+1
                    record.loc[record['symbol']== sec,'last_buy_price']=current_price                            # 加仓价格记录         
    #策略部分(离场：止损或止盈)：当价格相对上个买入价下跌 2ATR时（止损）或当股价跌破10日唐奇安通道（止盈），清仓离场 
                elif current_price< low[-int(DC_range/2):-1].min() or current_price < (last_price - 2*atr) or current_price < ma10[-1]:
                    long_position=account.get_positions().get(sec, dict()).get('long_amount', 0)
                   
                    order_id=account.order(sec,-long_position,'close')                              # 清仓离场
                    # _order=account.get_order(order_id)
                    # print(context.now)
                    
                    # if _order.state != OrderState.FILLED :
                    #     time.sleep(15)
                    # assert _order.state == OrderState.FILLED
                    # if _order.state == OrderState.FILLED :
                    record = record[record['symbol']!=sec]       # 将卖出期货的记录清空   
         
```
