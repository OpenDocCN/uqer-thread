# 行业集中度分析程序

### part1 程序说明
创建时间：2020-10-19 

功能：本程序主要用于分析各行业的集中度。
分析方法：通过2个指标分析行业集中度。   
1. 行业前3位的上市公司市值占全行业的上市公司的比重
2. 市值第一的公司市值与本行业市值中位数之比
3. 上述2个指标越大表示行业集中度越高。

版本更新：   
+ 20201020：优化了分组聚合运算，采用.agg([(列名,函数名),...])形式，更简洁优雅，效率更高   

### part2 全市场行业分析

```python
import numpy as np
import pandas as pd
import CAL.PyCAL as cal

import lib.mylib as ml   #导入自定义函数库

# 环境参数设置
DataAPI.settings.cache_enabled = False #调试时设为False。True:开启缓存，当前Notebook所有DataAPI数据都会在内存进行缓存
pd.options.display.max_columns = 100  #显示所有列

# 函数定义
def get_date(today=''):
    '''
    根据输入，设置日期为今天或某个指定的日期
    params:
        today->str:日期字符串，格式“YYYY-MM-DD”，默认为空，表示取今天日期
    return:
        tradeDate->cal.Date object:CAL日期类型的对象
    '''
   
    if today=='':
        today=cal.Date.todaysDate() #今日日期
    else:
        today=cal.Date.strptime(today,dateFormat='%Y%m%d') #转换成CAL的date对象
    calender=cal.Calendar('China.SSE') #获取上交所的交易日日历
    span = cal.Period('1D') #时间间隔
    tradeDate=calender.adjustDate(today-span,cal.BizDayConvention.Preceding) #根据交易所日历求前一个交易日
    tradeDate=tradeDate.strftime('%Y%m%d')
    return tradeDate

def get_mvConc01(gp):
    '''
    计算行业集中度指标
    params:
        gp->Series: 分组对象
    return
        ->float:返回包含计算好的行业集中度指标的df对象
    '''
    gp=gp.sort_values(ascending=False)   #排序，从大到小
    return gp[:3].sum()/gp.sum()   #计算前3位之和与总和的比值

def get_mvConc02(gp):
    '''
    计算行业集中度指标
    params:
        gp->Series: 分组对象
    return
        ->float:返回包含计算好的行业集中度指标的df对象
    '''
    return gp.max()/gp.median()   #计算最大值与中位数的比值

tradeDate=get_date('')  #取当天日期或指定日期
print("数据采样日期：{}".format(tradeDate))

universe=set_universe('A') #所有A股代码的列表
df=DataAPI.EquIndustryGet(secID=universe,industryVersionCD=u"010303",
                          field=u"secID,isNew,industryID2,industryName2",pandas="1") #取行业分类数据
df=df[df.isNew==1][['secID','industryID2','industryName2']]  #保留最新分类数据
df_ind=df.dropna().drop_duplicates()   #去空去重

df=DataAPI.MktEqudGet(secID=universe,tradeDate=tradeDate,
                      field=u"secID,secShortName,tradeDate,marketValue",pandas="1")  #取总市值数据
df_mv=df.dropna().drop_duplicates()

df=pd.merge(df_ind[['secID','industryID2']],df_mv[['secID','marketValue']],
            how='inner',left_on='secID',right_on='secID')   #合并分类和市值数据

gp=df.groupby('industryID2')      #按行业代码分类
func=[('size','size'),('mvConc01',get_mvConc01),('mvConc02',get_mvConc02)]    #对分组进行计算的函数列表
df_conc=gp['marketValue'].agg(func)   #运用函数列表计算
df_conc=pd.merge(df_ind[['industryID2','industryName2']].drop_duplicates(),df_conc,
                      how='inner',left_on='industryID2',right_index=True)        #添加分类名称
df_conc=df_conc.sort_values('industryID2').reset_index(drop=True)        
df_conc.sort_values('mvConc02',ascending=False).reset_index(drop=True)      #结果展示   
```
