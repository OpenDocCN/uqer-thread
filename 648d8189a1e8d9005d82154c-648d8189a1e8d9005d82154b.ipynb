# 数据处理：日期数据

```python
# 加载需要用的包
import numpy as np
import pandas as pd

# 得到每月月末的交易日日期
data = DataAPI.TradeCalGet(exchangeCD=u"XSHG", beginDate='20140101', endDate='20240101', field=['calendarDate','isWeekEnd','isMonthEnd'], pandas="1")
data = data[data['isMonthEnd'] == 1]
date_list = map(lambda x: x[0:4]+x[5:7]+x[8:10], data['calendarDate'].values.tolist())
date_list

#取出3，4，8，10月月末的交易日日期refresh_date
refresh_date=np.zeros(len(date_list)/3)                 #一年只有1/3个月会调仓
for i in range(1,len(date_list)+1):                     #以index除以12的余数来判断当前index所对应的是几月
    if i%12 == 3:                                       #余数为3，表示当前index对应的月是三月，把它存起来，放进refresh_date里，下面一样
        refresh_date[int(i/12)*4+0] = date_list[i-1]
    if i%12 == 4:
        refresh_date[int(i/12)*4+1] = date_list[i-1]
    if i%12 == 8:
        refresh_date[int(i/12)*4+2] = date_list[i-1]
    if i%12 == 10:
        refresh_date[int(i/12)*4+3] = date_list[i-1]

refresh_date=list(refresh_date)        #把refresh_date存为list
refresh_date=[str(int(x)) for x in refresh_date]  
refresh_date.sort(reverse=True)       #排序是为了之后会用到，这里是否排序并不影响策略的构建
```

```python
def stock_selection_basic(context):
    '''
    买入股东人数明显减少的股票组合
    '''

    # 基础数据获取（通用数据）：投资者账户，可供投资股票，价格数据，持仓数据。账户余额数据
    today = context.current_date                                                        # 获取当前日期
    account = context.get_account('fantasy_account')                                    # 获得投资者的账户（fantasy_account）
    current_universe = context.get_universe(exclude_halt = True)                        # 获取当前除停牌外的所有可供投资股票（universe）
    security_position = account.get_positions()                                         # 字典型数据，上一个K线结束后的有效证券头寸，且持仓数量大于0的证券及其头寸
    cash = account.cash                                                                 # 获取股票账户可用于投资的现金额度
    previous_date = context.previous_date
    previous_date = previous_date.strftime('%Y%m%d')
    startday = str(int(previous_date[:4]) - 1) + previous_date[-4:]


    # 数据处理部分：（1）整理存储原始数据；（2）进行平均持股变化率的计算; (3) 选出最近平均持股变化率大于20%的股票，且股东人数小于2万人
                                                      
    stkdata = {}                                                                        # 用于存原始数据的字典，key就是股票的secID,key所对应的是一个dataframe
    stkdatapro={'symbol': [], 'sec_hold_ratio': [], 'securityNum': []}                  # 设置处理后的数据字典型变量，以股票代码为key, 股东户数变化率为Value


    for sec in current_universe:                                                        # 以secID对研究范围内的股票进行循环
        #注意这里输入是ticker
        holdsharesdata = DataAPI.EquShareholderNumGet(secID=u"", ticker = sec[0:6], beginDate = startday, endDate=u"", beginPublishDate=u"", endPublishDate = previous_date, field=['publishDate', 'securityNum', 'avgHoldNum'], pandas="1")

        try:
            # （1）整理存储原始数据；
            stkdata[sec] = holdsharesdata                                                # 以secID为key,把该只股票的相关数据赋给它
            stkdata[sec]['sec_num_ratio'] = stkdata[sec]['securityNum'] / stkdata[sec]['securityNum'].shift(-1) - 1
            stkdata[sec]['sec_hold_ratio'] = stkdata[sec]['avgHoldNum'] / stkdata[sec]['avgHoldNum'].shift(-1) - 1
            #stkdata[sec].dropna(inplace = True)
            stkdata[sec].sort_values('publishDate', ascending = False, inplace = True)
            # （2）进行总户数变化率的计算;
            #ratio_change = float((stkdata[sec]['sec_hold_ratio'].iloc[0] + stkdata[sec]['sec_hold_ratio'].iloc[1])/2)
            ratio_change = float(stkdata[sec]['sec_hold_ratio'].iloc[0])
            stkdatapro['symbol'].append(sec)
            stkdatapro['sec_hold_ratio'].append(ratio_change)
            stkdatapro['securityNum'].append(float(stkdata[sec]['securityNum'].iloc[0]))

        except:
            print('This code is %s not vaild'%sec)

    # （3）选出最近平均持股变化率大于20%的股票，且股东人数小于2万人
    select_sec = pd.DataFrame(stkdatapro)
    select_sec.dropna(inplace = True)
    final_sec = select_sec.loc[(select_sec['securityNum'] < 20000) & (select_sec['sec_hold_ratio'] > 0.2) , :]
    sec_list = final_sec['symbol'].tolist()

    return sec_list
```
