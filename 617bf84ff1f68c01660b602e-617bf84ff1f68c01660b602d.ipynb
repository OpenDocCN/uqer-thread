```python
# 各个量化平台有自己的证券数据库编码
# 先 通过其他软件确定 50ETF、300ETF、可转债ETF、互联网ETF、光伏ETF 的证券代码
# 注意 调取方法 的 参数 和可选返回值的 书写方法
# dfETFSecID = DataAPI.SecIDGet(ticker=u"510050,510300,511380,517200,515790",field=u"secID,ticker,secShortName",pandas="1")
dfETFSecID = DataAPI.SecIDGet(ticker=u"510050,510300,511380,515790",field=u"secID,ticker,secShortName",pandas="1")

# print(dfETFSecID)
# 返回的数据类型为dataframe(数据框)

# 基于pandas(数据框操作工具组件) 提取所需的数据
import pandas as pd
# 指定内容筛选
# 在dfETFSecID筛选secID列中 带有".XSHG"的行   if ".XSHG" in x.secID  为筛选条件
dfTrueFalse = dfETFSecID.apply(lambda x: True if ".XSHG" in x.secID else False,axis=1)
# print(dfTrueFalse)
# 应用筛选 - 把 具有相同行名的TF数据框传入 执行筛选操作的数据框
dfNeed = dfETFSecID[dfTrueFalse]
# print(dfNeed)

# # 符合筛选的一种实现方法 单个 条件时 可以 省略  True if 和 else False
# dfTrueFalse = dfNeed.apply(lambda x: "华泰" in x.secShortName,axis=1)
# dfNeed = dfNeed[dfTrueFalse]
# print(dfNeed)

# # 将数据框的某列数据最终变成一串字符串
# # 提取某列数据   （链式方法调用）
# dfI = dfNeed.loc[:,'secID']
# print(dfI)
# print(type(dfI))
# # 转换为数组  （链式方法调用）
# data = dfI.values
# print(data)
# print(type(data))
# # 转换为列表 （链式方法调用）
# dataList = data.tolist()
# print(dataList)
# print(type(dataList))
# # 转换为字符串  （字符串，入参方法调用） 
# SecIDdataStr = ','.join(dataList)
# print(SecIDdataStr)

# 也可以把所有代码链式放在一起使用一行
SecIDdataStr = ','.join(dfNeed.loc[:,'secID'].values.tolist())
# print(SecIDdataStr)

# 获取期间单位净值
dfNAV = DataAPI.FundNavGet(secID=SecIDdataStr,
                       beginDate=u"20200101",
                       endDate=u"20201231",
                       field=u"secID,NAV,endDate",
                       pandas="1")
# print(dfNAV)

# 只有 686 行 并非5的倍数 一定存在缺失值(行数缺失) 
# ---循环的一般书写方法
# 先按secID 拆分数据 
# 先确定循环变量 i 这里要考察具体哪只合约 缺失 ，缺失成何种样子
secIDList = dfNeed.loc[:,'secID'].values.tolist()

# secIDI = secIDList[0]
# print(secIDI)
# # 筛选的简易写法，只限 简单的不等式条件
# dfNAVI = dfNAV[dfNAV.secID==secIDI]
# # print(dfNAVI)
# print(dfNAVI.shape)

for secIDI in secIDList:
    # print(secIDI)
    dfNAVI = dfNAV[dfNAV.secID==secIDI]
    # print(dfNAVI)
    # print(dfNAVI.shape)

    
# 解决方法 替换标的 切换时间
# 获取期间单位净值
dfNAV = DataAPI.FundNavGet(secID=SecIDdataStr,
                       beginDate=u"20200101",
                       endDate=u"20210430",
                       field=u"secID,NAV,endDate",
                       pandas="1")
# print(dfNAV)
# for secIDI in secIDList:
#     print(secIDI)
#     dfNAVI = dfNAV[dfNAV.secID==secIDI]
#     # print(dfNAVI)
#     print(dfNAVI.shape)
    
# 绘图数据整理
# 需要将4列数据横向放在一张数据框内，也即 endDate 作为 行名 secID作为 列名 NAV净值作为 数据 的数据框
# 依旧是循环

dfList = []
for secIDI in secIDList:
    # secIDI = secIDList[0]
    # print(secIDI)
    dfNAVI = dfNAV[dfNAV.secID==secIDI]
    # print(dfNAVI)

    dfNAVI = dfNAVI.loc[:,['NAV','endDate']]
    dfNAVI = dfNAVI.rename(columns ={'NAV':secIDI+'_NET'})
    # print(dfNAVI)
    dfList.append(dfNAVI)

# 进行拼接
for i in [1,2,3]:
    dfList[0] = pd.merge(dfList[0],dfList[i],how='left',on='endDate')
# print(dfList[0])
# 列表解析 简易的循环写法
columnsList = ['endDate'] + [secIDI+'_NET' for secIDI in secIDList]

# 利用选列的方式，达成整理字段顺序
dfResult = dfList[0].loc[:,columnsList]
# print(dfResult)

# 绘图 导入基础的绘图库
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(12, 7))
colorList = ["red","b","green","#E2D2A1","pink","#FFDD44",(1.0,0.2,0.3),"blue"]
linestyleList = ["solid","dashed",'-.',':','--']
# 一种同步循环的实现
for secIDI, color, linestyle in zip(secIDList, colorList[:len(secIDList)], linestyleList[:len(secIDList)]):
    # print(secIDI+'_NET')
    # print(color)
    # print(linestyle)
    # ax.plot(dfResult.endDate.values, dfResult.loc[:,secIDI+'_NET'].values, color=color,linestyle=linestyle)
    ax.plot(range(len(dfResult)), 
            dfResult.loc[:,secIDI+'_NET'].values,
            color=color,
            linestyle=linestyle,
            label = secIDI)


DateLabelsList = dfResult.endDate.values.tolist()

# 图例位置
plt.legend(loc='upper left')

# 这里设置的是x轴点的位置（40设置的就是间隔了）
xticks=list(range(0,len(DateLabelsList),40)) 
#这里设置X轴上的点对应在数据集中的值（这里用的数据为DateLabelsList）
xlabels=[DateLabelsList[x] for x in xticks] 
xticks.append(len(DateLabelsList))
xlabels.append(DateLabelsList[-1])
ax.set_xticks(xticks)
ax.set_xticklabels(xlabels, rotation=40)

plt.show()


```
