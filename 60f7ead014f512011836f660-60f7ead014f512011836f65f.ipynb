4.1 datetime构造
时间类型 
date      日期（年、月、日）
time      时间（时、分、秒、毫秒）
datetime  日期和时间
timedelta 两个datetime的差（日、秒、毫秒）

```python
#4.1datetime构造
import datetime
date=datetime.date(2018,3,12)
print(date)

print(date.year)
print(date.month)
print(date.day)

time=datetime.time(9,10,34)
print(time)

print(time.hour)
print(time.minute)
print(time.second)

print'现在的时间是：'
now=datetime.datetime.now()#年、月、日、小时、分钟、秒、毫秒
print(now)

print'时间差：'
delta=now-datetime.datetime(2020,1,1,1,1)#两个时间的差
print(delta)

print'从现在往后推10天'
time=now+datetime.timedelta(10)
print(time)
```

```python
#4.2数据转换
from datetime import datetime 
stamp=datetime(2018,5,12)
print(stamp)

#时间序列转字符串
print(str(stamp))
print(stamp.strftime('%Y/%m/%d'))#自定义时间的格式
print(stamp.strftime('%Y-%m'))

print(stamp.strftime('%W'))#输出该日期为当年的第几周

#字符串转时间序列
value='2018-4-12'
datetime.strptime(value,'%Y-%m-%d')
```

```python
#4.3时间序列基础（构造，索引与切片）
from datetime import datetime
import numpy as np
import pandas as pd

#时间序列的构造
date=[datetime(2018,4,1),datetime(2018,4,5),
     datetime(2018,4,7),datetime(2018,4,9),
     datetime(2018,4,10),datetime(2018,4,15)]
s=pd.Series(np.arange(6),index=date)
print(s)

print(type(s))
print(s.index)

print(s[::2])#括号内的意思是从开始到结束，步长为2，所以输出的是第1、3、5行的内容

print(s+s[::2])#解释不出来，赋值？？？

print(s.index[0])#输出序列s中第0行的索引

#时间序列的索引和切片
date=[datetime(2018,4,1),datetime(2018,4,5),
     datetime(2018,4,7),datetime(2018,4,9),
     datetime(2018,4,10),datetime(2018,4,15)]
s=pd.Series(np.arange(6),index=date)
print(s)

#索引
print(s[0])          #时间序列的第0行的值
print(s[1:3])        #时间序列的第1-2行的值
print(s['20180401']) #输出20180401是时间序列的第几行
print(s['2018/4/5']) #输出2018/4/5是时间序列的第几行
#切片
print(s['2018/4/5':'2018/4/11'])
print(s[datetime(2018,4,7):])

date2=[datetime(2018,4,1),datetime(2018,4,5),
      datetime(2018,4,13),datetime(2018,4,27),
      datetime(2018,9,4),datetime(2018,9,8),
      datetime(2018,9,12),datetime(2018,9,23),
      datetime(2019,3,12),datetime(2019,3,27),
      datetime(2019,6,7),datetime(2019,6,17)]
long_s=pd.Series(np.arange(12),index=date2)
print(long_s)

print(long_s['2019'])    #寻找2019年的数据
print(long_s['2019/6'])  #寻找2019年6月的数据

date3=pd.DatetimeIndex(['2018/4/14','2018/4/14',
                       '2018/5/23','2018/5/23',
                       '2018/6/13','2018/6,13'])
dup_s=pd.Series(np.arange(6),index=date3)
print(dup_s)

print(dup_s.index.is_unique)#判断dup_s序列的索引是否是独一无二的
print(dup_s.groupby(level=0).mean())#同一日期下求均值
print(dup_s.groupby(level=0).sum())#同一日期下求和
```

4.4日期
别名
D              每日历日
B              每工作日
H              每小时
T或者min       每分钟
S              每秒
M              每月最后一个日历日
BM             每月最后一个工作日
A-JAN、A-FEB...每年指定月份的最后一个日历日

```python
#4.4日期
index=pd.date_range('2018/4/1','2018/5/30')#date_range构造长段时间序列
print(index)

index=pd.date_range('2018/4/1','2018/12/31',freq='M')#指定频率为月，M每月最后一个日历日
print(index)

index=pd.date_range('2018/4/1',periods=20)#从2018年4月1号开始的20天
print(index)

index=pd.date_range(end='2018/6/1',periods=20)#从2018年6月1号往前推20天
print(index)

index=pd.date_range(start='2018/6/1 15:11:34',periods=10)
print(index)

index=pd.date_range(start='2018/6/1 15:11:34',periods=10,normalize=True)#后面的时候不输出
print(index)
```

```python
#4.5频率与移动
#频率
index=pd.date_range(start='2018/4/1',periods=10,freq='4H')
print(index)

index=pd.date_range(start='2018/4/1',periods=10,freq='2H20min38S')
print(index)

#频率
date4=pd.date_range('2018/4/1',periods=5)
s=pd.Series(np.arange(5),index=date4)
print(s)
print(s.shift(2))#序列往下移动2位，缺失值用nan代替
print(s.shift(-2))#序列往上移动2位，缺失值用nan代替
print(s.shift(2,freq='D'))#给定频率之后，就是讲index进行平移
print(s.shift(2,freq='M'))
```

#4.6频率转换与重采样
参数             使用说明
freq             转换频率
axies=0          重采样的轴
closed='right'   在降采样中，各时间段哪端是闭合的
lable='right'    在降采样中，如何设置聚合值的标签
loffset=None     设置时间偏移量
kind=None        聚合到时期或时间戳，默认为时间序列的索引类型
convention=None  升采样所采用的约定（start或end）,默认为end

```python
#4.6频率转换与重采样
#4.6.1重采样
import numpy as np
import pandas as pd 

date=pd.date_range(start='2018/4/1',periods=100,freq='D')
s=pd.Series(np.arange(100),index=date)
print(s.head(10))#只看前10个
print(s.resample('M').mean())#日期显示的是每月最后一天，数据显示每个月的均值

#4.6.2降采样(降低采样频率)
date=pd.date_range(start='2018/4/1',periods=12,freq='D')
s=pd.Series(np.arange(12),index=date)
print(s)
#采样数据位5天求和，左开右闭，用右边的日期作为标签
s.right=s.resample('5D',closed='right',label='right').sum()
print(s.right)
#采样数据位5天求和，左开右闭，用右边的日期作为标签
s.left=s.resample('5D',closed='left',label='left').sum()
print(s.left)
#采样数据位5天求和，左开右闭，用右边的日期作为标签之后往前偏移一天
s.right=s.resample('5D',closed='right',label='right',loffset='-1D').sum()
print(s.right)

#4.6.3升采样(提高采样频率，并进行填充数据)
from datetime import datetime
date=[datetime(2018,4,3),datetime(2018,4,13)]
s=pd.Series([2,5],index=date)
print(s)

print(s.resample('D').ffill())#前面的数据往后填充
print(s.resample('D').bfill())#后面的数据往前填充
print(s.resample('D').ffill(3))#前面的数据往后填充3位
```
