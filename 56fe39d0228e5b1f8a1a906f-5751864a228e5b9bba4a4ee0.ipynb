```python
import pandas as pd
import numpy as np
import datetime as dt   
 
    
begin_date = u'20060601'
end_date = u'20160601'
field = ['turnoverValue','turnoverVol','tradeDate','closePrice']
#print pd.__version__
idx_info=DataAPI.MktIdxdGet(ticker='000001',beginDate=begin_date,endDate=end_date,field=['turnoverValue','turnoverVol','tradeDate','closeIndex'])

#idx_info = DataAPI.MktEqudGet(ticker=u"000001",beginDate=begin_date,field=['turnoverValue','turnoverVol','tradeDate','closePrice'])

# 前复权
#idx_info = DataAPI.MktEqudAdjGet(ticker=u"000001",beginDate=begin_date,endDate=end_date,field=field)
# 后复权
#idx_info = DataAPI.MktEqudAdjAfGet(ticker=u"000001",beginDate=begin_date,endDate=end_date,field=field)

idx_info['5_avarage'] = pd.rolling_sum(idx_info.turnoverValue, 5) / pd.rolling_sum(idx_info.turnoverVol, 5)
idx_info['10_avarage'] = pd.rolling_sum(idx_info.turnoverValue, 10) / pd.rolling_sum(idx_info.turnoverVol, 10)
idx_info['month_avarage'] = pd.rolling_sum(idx_info.turnoverValue, 21) / pd.rolling_sum(idx_info.turnoverVol, 21)
idx_info['quarter_avarage'] = pd.rolling_sum(idx_info.turnoverValue, 63) / pd.rolling_sum(idx_info.turnoverVol, 63)
idx_info['halfyear_avarage'] = pd.rolling_sum(idx_info.turnoverValue, 125) / pd.rolling_sum(idx_info.turnoverVol, 125)
idx_info['year_avarage'] = pd.rolling_sum(idx_info.turnoverValue, 125) / pd.rolling_sum(idx_info.turnoverVol, 125)
idx_info.closeIndex = idx_info.closeIndex / 300 + 0.0

del idx_info['turnoverValue']
del idx_info['turnoverVol']
idx_info.index = idx_info.tradeDate.apply(lambda x: dt.datetime.strptime(x, '%Y-%m-%d'))
del idx_info['tradeDate']

print idx_info.head()

idx_info.plot(figsize=(20,15))















```
