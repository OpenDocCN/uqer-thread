# Introduction

一张在各大《投资学》中通常都会有的图，亦即基金月度收益率分布。换句话说，如果我们去看如蚂蚁聚宝，或者基金说明书，一般都会说，股票基金风险大于混合型基金大于债券基金，那么这是什么意思。从分布的角度，股票基金收益率比较分散，从小（可能为负）到大比较宽，分布的峰比较平；而债券基金通常分布比较集中，峰比较陡峭。混合基金的分布当然在股票基金和债券基金之间。于是和 $ \sigma $ 作为风险衡量相对称。

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn
```

```python
# uqer.datayes.com的基金分类：E-股票型，H-混合型，B-债券型，SB-短期理财债券，M-货币型

# subplots的使用，返回一个figure和ax对象，其中此时的ax是[[ax0,ax1],[ax2,ax3]]
fig, ax = plt.subplots(3, 1, sharex=True)

# 动态调整fig的大小
fig.set_size_inches(12,24)

# category_list = ['E','H','B','M']
category_list = ['E','H','B']
for num,each_type in enumerate(category_list): # enumerate使得返回序号和内容
    
    fundlist = DataAPI.FundGet(etfLof=u"",secID=u"",ticker=u"",category=each_type,idxID=u"",idxTicker=u"",listStatusCd=u"",operationMode=u"",field=u"",pandas="1")
    ll = []
    if each_type in ['E','H','B']:
        for secID in fundlist['secID']:
            df = DataAPI.FundNavGet(dataDate=u"",secID=secID,ticker=u"",beginDate=u"",endDate=u"",field=u"",pandas="1")
            df = df[['endDate','ADJUST_NAV']]
            df = df.rename(columns={'ADJUST_NAV':secID})
            df['endDate'] = pd.to_datetime(df['endDate'])
            df.set_index('endDate',inplace=True)
            ll.append(df)
    elif each_type in ['M']:
        try:
            raise NotImplementedError
            for secID in fundlist['secID']:
                pass
        except NotImplementedError:
            pass
    full_df = pd.concat(ll,axis=1)
    kk = (full_df.resample('m').last() - full_df.resample('m').last().shift(1))/full_df.resample('m').last().shift(1) *100
    kk = kk.values.flatten()
    kk = kk[~np.isnan(kk)]
    ss = pd.Series(kk)
    # 不用plt.hist，因为涉及到uqer.datayes.com这里会直接输出
    # stats,bins,patches = plt.hist(ss,bins=200,normed=False)
    stats,bins = np.histogram(ss,200)
    ax[num].bar((bins[1:]+bins[:-1])/2,stats,width=0.8)
    ax[num].set_title("Fund_"+each_type)
    ax[num].set_xlim([-50,50])

plt.show()
```
