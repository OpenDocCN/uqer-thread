# 广播
两个数组的维度不同，numpy可以通过广播的方式实现对形状不相似的数组进行操作。较小的数组会广播到较大的数组中。
reference reading: https://docs.scipy.org/doc/numpy/user/basics.broadcasting.html

```python
# e.g
# 1. outer product
# 2. add a vector to each row of a matrix
a = np.arange(12).reshape(4, 3)
b = np.array([1, 0, 1])
y = np.empty_like(a)
for i in range(4):
    y[i, :] = a[i, :] + b
# 3. add a vector to each column of a matrix
# 4. multiply a matrix by a constant
```

```python
print a
print b

print y

print a + b
```

```python
a = np.arange(12).reshape(3, 4)
b = np.array([1, 0, 1])

print a
print b

print (a.T + b).T
```
