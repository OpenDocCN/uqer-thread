## **2.多因子—排列组合**

---

#### **2.1 理论准备**

* 本篇受启发于国信证券“基于A股市场选股因子边际效用和有效分散的动态分度动量策略”（下载地址见文末）。

* 在该研报中，研究人员选择了23个因子，通过穷举的方法组合构建了23个单因子模型、253个双因子模型、347个三因子模型、520个四因子模型以及300个五因子模型，通过分析他们的历史表现，得出因子边际效用递减的结论。并提出动态调整区分度的半衰期策略。

##### **本篇的目的：**

- 利用报告中因子区分度与贡献度的概念，对列举的因子进行筛选，选出排名最前的5个因子。

- 以选取的5个因子为例，进行排列组合，构建这些因子的单因子模型、双因子模型以及三因子模型，比较分析它们的回测情况。

#### **2.2 因子贡献度**

* 因子贡献度的思路是将股票池中的股票按因子进行排名，分别选出排名靠前的20%和排名靠后的20%股票构成两个组合。两个组合的收益率相差越大，则说明该时点此因子的强度越高。

* 根据这个思路，我们选取以下15个因子，考察它们的贡献度con。

###### **因子有：**

>1.估值指标 
* ###### PE：指定日市盈率 
* ###### PB：指定日市净率
* ###### PCF：市现率

>2.规模指标 
* ###### LCAP：对数市值 
* ###### ETOP：收益市值比

>3.成长指标：
* ###### TotalAssetGrowRate：总资产增长率，最新一期财报的总资产增长率 
* ###### NetAssetGrowRate：净资产增长率
* ###### OperatingRevenueGrowRate：营业收入增长率
* ###### NetProfitGrowRate: 净利润增长率

>4.财务指标 
* ###### ROE：净资产收益率（最近一年滚动处理
* ###### ROA：资产回报率
* ###### QuickRatio：速动比率
 
>5.技术指标 
* ###### VOL60：60日平均换手率 
* ###### RSI：相对强弱指标
* ###### Volatility：换手率相对波

###### **股票池：**

>这里选择2011年12月沪深300指数的成分股作为我们固定的股票池。避免在计算因子贡献度时，因股票池成分发生变化，而使对照组产生更多的变量。

###### **计算时间：**

>2012年1月1日——2012年12月31日

```python
import pandas as pd
from pandas import DataFrame
import numpy as np
from CAL.PyCAL import *
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('white')
# import datetime as dt
```

```python
# 获取11年12月份沪深300的成分股作为股票池
universe = DataAPI.IdxCloseWeightGet(ticker=u"000300",beginDate=u"20111201",endDate=u"20111231",field=u"consID",pandas="1").consID.tolist()

# 因子池
factors = ['PE','PB','PCF', 'LCAP','ETOP', 'TotalAssetGrowRate','NetAssetGrowRate','OperatingRevenueGrowRate','NetProfitGrowRate','ROE','ROA','QuickRatio','VOL60','RSI','Volatility']

# --------------------- 数据准备  -------------------- #

ret = {}
for stk in universe:
    tmp = DataAPI.MktEqudGet(secID=stk,beginDate=u'20120101',endDate=u'20121231',field=[u'tradeDate', u'ticker',u'closePrice'],pandas="1")    
    # 计算股票池的股票在12年的全年的收益率
    ret[stk] = (tmp['closePrice'].iloc[-1] - tmp['closePrice'].iloc[0]) / tmp['closePrice'].iloc[0]     
ret_pd = pd.DataFrame()
ret_pd['secID'] = ret.keys()
ret_pd['return'] = ret.values()
R_data = ret_pd.set_index('secID')

for stk in universe:
    tmp = DataAPI.MktStockFactorsDateRangeGet(secID=stk,beginDate=u"20120101",endDate=u"20121231",field=['secID'] + factors,pandas="1").set_index('secID')
    # 计算股票池的股票各因子在12年的全年的表现，这里使用各因子在12年的全年均值
    ret[stk] = tmp.mean()      
F_data = pd.DataFrame(ret).T

# 合并数据，获得全部原始数据
All_data = pd.merge(R_data, F_data, how='inner', left_index=True, right_index=True)   
All_data.to_csv('All_data.csv')
All_data.head()
```

##### 下面对数据进行整理，计算各个因子的贡献度。

```python
# 从csv读取DataFrame，会重新生成index，有什么解决方法求告知
All_data = pd.read_csv('All_data.csv').set_index('Unnamed: 0')

Return = All_data['return'].to_dict()  # 保存收益率数值做备用
ret = pd.DataFrame()
for fac in All_data.columns:
    tmp = All_data[fac].to_dict()
    p = standardize(neutralize(winsorize(tmp),'20120101'))    #数据标准化、中性化、去极值
    ret['secID'] = p.keys()
    ret['Return'] = Return.values()    # 循环中将return也做了标准化、中性化、去极值处理，所以这里用原来保存的Return做替换
    ret[fac] = p.values()
New_data = ret.sort('secID').set_index('secID').drop('return', axis=1)     
```

```python
# 求实际收益率排名前20%和后20%的股票的平均收益率
q = New_data.sort('Return',ascending=False).head(60)
R_head_mean = q['Return'].mean()
p = New_data.sort('Return',ascending=False).tail(60)
R_tail_mean = p['Return'].mean()

# 将股票池股票按照各因子大小进行排序，计算排序后前后20%股票在12年的收益率
con = {}
for fac in New_data.columns:
    tmp_head = New_data.sort(fac,ascending=False).head(60)
    fac_head_mean = tmp_head['Return'].mean()
    tmp_tail = New_data.sort(fac,ascending=False).tail(60)
    fac_tail_mean = tmp_tail['Return'].mean()
    con[fac] = (fac_head_mean - fac_tail_mean) / (R_head_mean - R_tail_mean)   # 因子分布两段的收益率差 / 实际收益率分布两端差值 = 因子贡献度
con_pd = pd.DataFrame()
con_pd['factors'] = con.keys()
con_pd['con_value'] = con.values()
con_pd.sort('con_value', ascending=False)   # 得到各因子贡献度数值排序
```

##### 对收益率贡献度最好的5个因子是RSI、LCAP、PB、ROE、NetProfitGrowRate。

#### **2.2 单因子模型**

---

##### 回测时间2013年1月1日——2015年8月1日；调仓周期：按月调仓；股票池：HS300；选股方案：每月初对上月底股票池中的股票按因子顺序进行排序，选出前20%的股票；权重分配：等权分配资金。

###### 夏普、信息比率等指标似乎都不是很理想，但没关系，我们只是一个简单的策略。

###### 下面利用 quantz 的回测框架，对所有单因子进行测试。

```python
# 可编辑部分与 strategy 模式一样，其余部分按本例代码编写即可

# -----------回测参数部分开始，可编辑------------
start = '2013-01-01'                       # 回测起始时间
end = '2015-08-01'                         # 回测结束时间
benchmark = 'HS300'                        # 策略参考标准
universe = set_universe('HS300')  # 证券池，支持股票和基金
capital_base = 10000000                      # 起始资金
freq = 'd'                                 # 策略类型，'d'表示日间策略使用日线回测，'m'表示日内策略使用分钟线回测
refresh_rate = 1                           # 调仓频率，表示执行handle_data的时间间隔，若freq = 'd'时间间隔的单位为交易日，若freq = 'm'时间间隔为分钟

# 构建日期列表
data=DataAPI.TradeCalGet(exchangeCD=u"XSHG",beginDate=u"20110801",endDate=u"20150801",field=['calendarDate','isMonthEnd'],pandas="1")
data = data[data['isMonthEnd'] == 1]
date_list = data['calendarDate'].values.tolist()

cal = Calendar('China.SSE')
period = Period('-1B')

factors = [['RSI'],['LCAP'],['PB'],['ROE'],['NetProfitGrowRate']]
# factors = ['RSI','LCAP','PB','ROE','NetProfitGrowRate']
# ---------------回测参数部分结束----------------


# 把回测参数封装到 SimulationParameters 中，供 quick_backtest 使用
sim_params = quartz.SimulationParameters(start, end, benchmark, universe, capital_base)
# 获取回测行情数据
idxmap, data = quartz.get_daily_data(sim_params)
# 运行结果
onefac_results = {}

# 调整参数
for fac in factors:
     # ---------------策略逻辑部分---------------- #
    refresh_rate = 1
    
    def initialize(account):
        pass
    
    def handle_data(account):
        today = account.current_date
        today = Date.fromDateTime(account.current_date)                   
        yesterday = cal.advanceDate(today, period)
        yesterday = yesterday.toDateTime()

        if yesterday.strftime('%Y-%m-%d') in date_list:
            FAC = DataAPI.MktStockFactorsOneDayGet(tradeDate=yesterday.strftime('%Y%m%d'),secID=account.universe,field=['secID'] + fac,pandas="1").sort(fac,ascending=False).head(60)
            stk_list = FAC['secID'].tolist()
                
            # 先卖出 
            sell_list = account.valid_secpos                
            for stk in sell_list:
                order_to(stk, 0)

            # 再买入
            buy_list = stk_list
            total_money = account.referencePortfolioValue
            prices = account.referencePrice 
            for stk in buy_list:
                if np.isnan(prices[stk]) or prices[stk] == 0:  # 停牌或是还没有上市等原因不能交易
                    continue
                order(stk, int(total_money / len(buy_list) / prices[stk] /100)*100)
        else:
            return
    # ---------------策略逻辑部分结束----------------

    # 把回测逻辑封装到 TradingStrategy 中，供 quick_backtest 使用
    strategy = quartz.TradingStrategy(initialize, handle_data)
    # 回测部分
    bt, acct = quartz.quick_backtest(sim_params, strategy, idxmap, data, refresh_rate=refresh_rate)

    # 对于回测的结果，可以通过 perf_parse 函数计算风险指标
    perf = quartz.perf_parse(bt, acct)
    
    # 保存运行结果
    tmp = {}
    tmp['bt'] = bt
    tmp['annualized_return'] = perf['annualized_return']
    tmp['volatility'] = perf['volatility']
    tmp['max_drawdown'] = perf['max_drawdown']
    tmp['alpha'] = perf['alpha']
    tmp['beta'] = perf['beta']
    tmp['sharpe'] = perf['sharpe']
    tmp['information_ratio'] = perf['information_ratio']
    
    for i in fac:
        onefac_results[i] = tmp
        print i
print 'done'
```

###### 回测成功，现在来看看各个单因子策略的alpha、年收益、beta、信息比率、最大回撤、夏普率、波动率

```python
One_fac = ['RSI','LCAP','PB','ROE','NetProfitGrowRate']
One_ret = pd.concat(pd.DataFrame.from_dict([onefac_results[fac]]) for fac in One_fac).drop('bt', axis=1)
One_ret.index = ['RSI','LCAP','PB','ROE','NetProfitGrowRate']
One_ret.columns = ['alpha','年收益','beta','信息比率','最大回撤','夏普率','波动率']
One_ret
```

```python
# 可编辑部分与 strategy 模式一样，其余部分按本例代码编写即可

# -----------回测参数部分开始，可编辑------------
start = '2013-01-01'                       # 回测起始时间
end = '2015-08-01'                         # 回测结束时间
benchmark = 'HS300'                        # 策略参考标准
universe = set_universe('HS300')  # 证券池，支持股票和基金
capital_base = 10000000                      # 起始资金
freq = 'd'                                 # 策略类型，'d'表示日间策略使用日线回测，'m'表示日内策略使用分钟线回测
refresh_rate = 1                           # 调仓频率，表示执行handle_data的时间间隔，若freq = 'd'时间间隔的单位为交易日，若freq = 'm'时间间隔为分钟

# 构建日期列表
data=DataAPI.TradeCalGet(exchangeCD=u"XSHG",beginDate=u"20110801",endDate=u"20150801",field=['calendarDate','isMonthEnd'],pandas="1")
data = data[data['isMonthEnd'] == 1]
date_list = data['calendarDate'].values.tolist()

cal = Calendar('China.SSE')
period = Period('-1B')

factors = [['RSI','LCAP'],['RSI','PB'],['RSI','ROE'],['RSI','NetProfitGrowRate'],['LCAP','PB'],['LCAP','ROE'],['LCAP','NetProfitGrowRate'],['PB','ROE'],['PB','NetProfitGrowRate'],['ROE','NetProfitGrowRate']]
# factors = ['RSI','LCAP','PB','ROE','NetProfitGrowRate']
# ---------------回测参数部分结束----------------


# 把回测参数封装到 SimulationParameters 中，供 quick_backtest 使用
sim_params = quartz.SimulationParameters(start, end, benchmark, universe, capital_base)
# 获取回测行情数据
idxmap, data = quartz.get_daily_data(sim_params)
# 运行结果
results = {}

# 调整参数
for fac in factors:
     # ---------------策略逻辑部分---------------- #
    refresh_rate = 1
    
    def initialize(account):
        pass
    
    def handle_data(account):
        today = account.current_date
        today = Date.fromDateTime(account.current_date)                   
        yesterday = cal.advanceDate(today, period)
        yesterday = yesterday.toDateTime()

        if yesterday.strftime('%Y-%m-%d') in date_list:
            FAC = DataAPI.MktStockFactorsOneDayGet(tradeDate=yesterday.strftime('%Y%m%d'),secID=account.universe,field=['secID'] + fac,pandas="1").dropna().set_index('secID')
            ret = pd.DataFrame()
            
            for i in FAC.columns.tolist():
                tmp = FAC[i].to_dict()
                p = standardize(neutralize(winsorize(tmp),'20130101'))
                ret['secID'] = p.keys()
                ret[i] = p.values()   
            ret = ret.set_index('secID')
            ret['score'] = ret.sum(axis=1)
            stk_list = ret.sort('score',ascending=False).head(60).index.tolist()

                
            # 先卖出 
            sell_list = account.valid_secpos                
            for stk in sell_list:
                order_to(stk, 0)

            # 再买入
            buy_list = stk_list
            total_money = account.referencePortfolioValue
            prices = account.referencePrice 
            for stk in buy_list:
                if np.isnan(prices[stk]) or prices[stk] == 0:  # 停牌或是还没有上市等原因不能交易
                    continue
                order(stk, int(total_money / len(buy_list) / prices[stk] /100)*100)
        else:
            return
    # ---------------策略逻辑部分结束----------------

    # 把回测逻辑封装到 TradingStrategy 中，供 quick_backtest 使用
    strategy = quartz.TradingStrategy(initialize, handle_data)
    # 回测部分
    bt, acct = quartz.quick_backtest(sim_params, strategy, idxmap, data, refresh_rate=refresh_rate)

    # 对于回测的结果，可以通过 perf_parse 函数计算风险指标
    perf = quartz.perf_parse(bt, acct)
    
    # 保存运行结果
    tmp = {}
    tmp['bt'] = bt
    tmp['annualized_return'] = perf['annualized_return']
    tmp['volatility'] = perf['volatility']
    tmp['max_drawdown'] = perf['max_drawdown']
    tmp['alpha'] = perf['alpha']
    tmp['beta'] = perf['beta']
    tmp['sharpe'] = perf['sharpe']
    tmp['information_ratio'] = perf['information_ratio']
    
    a = ''
    for i in fac:
        a += "+" + i
    results[a] = tmp
    print a,
print 'done'
```

## **存在的问题**

---

1. 由于我们在计算因子贡献度的时候，使用的因子值是2012年全年的平均值。这种计算方法对一些成长指标、财务指标是有一定的意义的，但是对于60日换手、RIS相对强弱指标等计算一年均值的参考意义就不是很大了，并且对净资产增长率、总资产增长率等指标来说，一年时间又是偏短的。

2. 在多因子组合进行调仓选股时，我们使用的方式是对股票池股票按两个因子去极值、取中值、标准化后的结果进行等权相加，选取排名前20%的股票作为buylist。然而在多因子模型当中，各因子不同的权重设置对结果影响非常大。
