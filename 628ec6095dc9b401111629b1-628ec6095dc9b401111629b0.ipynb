# MACD指标

MACD 分三部分：DIF、DEA、红绿柱(MACD)。
## DIF 
1.DIF=EMA12-EMA26

MA是均线，过去一段时间的收盘价均值，

EMA是加权的均线，提高近一段时间收盘价的权重。

EMA提近期权重会让均线更敏感，也提高了出错概率。

那DIF就很简单了，近期上涨，DIF变大，近期下跌，DIF变小。

DIF上穿0轴，就是12均线与26均线金叉，反之死叉。金叉死叉跟10周线、30周线一样的，金叉只是最近在涨，未来仍不确定。DIF周期更短，也就更敏感些。DIF为正时，值越大，就是12均线远大于26均线，意味着短期加速上涨，反之加速下跌。
## DEA 
2.DEA: DIF的MA这就像上篇所说的K线在均线之上。当DIF在DEA之上时，行情不错。DEA是DIF的均线，那DEA就会更迟钝点，高点、低点都会比DIF晚。当DIF为正时，DIF与DEA金叉，均值落后，意味着近期加速上涨，死叉则是上涨势头减弱。

## MACD
3.红绿柱: (DIF-DEA)*2值为负是绿柱，为正是红柱。红柱越长，就是上面说的，DIF为正，DIF与DEA金叉，加速上涨，金叉的点就是由绿柱转红柱。绿柱越长就是下跌加速。
