```python
#分析目标：周一涨幅和本周涨幅的相关性
#标的：指数，HS300，ZZ500，CYB
#时间:20130101-20140701,20140701-20151231

import numpy as np
import pandas as pd
import CAL.PyCAL as pycal
import matplotlib.pyplot as plt

start='20140701'
end='20150615'
all_stock_prices=pd.DataFrame()
print '### start '
stock_prices=None
fields=['secID','tradeDate','closePrice','accumAdjFactor','highestPrice','lowestPrice']
fieldsIndex=['ticker','tradeDate','closeIndex','highestIndex','lowestIndex','openIndex']
for stockSecID in ['399006','000016','000001']:#set_universe('HS300'),399006,000016,000001
    #获取价格数据
    stock_prices=None
    try:  
        stock_prices=DataAPI.MktIdxdGet(indexID=u"",ticker=stockSecID,tradeDate='',beginDate=start,endDate=end,field=fieldsIndex,pandas="1")
        stock_prices.rename(columns={'ticker': 'stockSecID','closeIndex':'closePrice','highestIndex':'highestPrice','lowestIndex':'lowestPrice','openIndex':'openPrice'}, inplace=True)
    except Exception ,e:
        print e
        
    if stock_prices is None or stock_prices.size==0:
        print 'stock_prices is null:',stockSecID,start,end
        continue
    #新增列，表示周几，函数判断周一周五周一0，周五为4，删除非周一，周五以及重复的004,044等非法情况（004去处开始的0,044去除结尾的4）
    stock_prices['weekday'] = pd.DatetimeIndex(stock_prices['tradeDate']).weekday
    stock_prices=stock_prices[stock_prices['weekday'].isin([0,4])]
    
    #周五的收盘价前移到周一同行
    stock_prices['nextClosePrice']=np.append(np.array(stock_prices['closePrice'][1:]),None)
    stock_prices['nextWeekday']=np.append(np.array(stock_prices['weekday'][1:]),None)
    
    #删除中间和结尾的无用数据
    stock_prices=stock_prices[np.logical_and(stock_prices['weekday']==0,stock_prices['nextWeekday']==4)]
    #统计，周一上涨/跌，本周上涨/跌，共2*2=4种情况，以及周一上涨，本周涨幅大于周一涨幅(就是后四天也在涨)的统计2*2=4种情况
    #extend：是否可以用周一上午涨跌预测本周？，周一上午涨跌和周一涨跌是否相关？
    
    #跌0涨1，周一涨本周涨则为11=3（二进制），同理跌跌=00=0，跌涨=01=2，涨跌=10=3，看不懂的话可自行循环方法统计
    stock_prices['result1']=(stock_prices['closePrice']>stock_prices['openPrice']).apply(lambda x:1 if x==True else 0)*2+\
                (stock_prices['nextClosePrice']>stock_prices['openPrice']).apply(lambda x:1 if x==True else 0)

    tmpResult=stock_prices['result1'].value_counts().sort_index()/stock_prices.shape[0]#各情况概率
    print '周一和本周相关性',stockSecID,tmpResult[0]+tmpResult[3]#00和11表示周一和本周一致的概率
    
    #周一和本周后四天相关性(认为周一尾盘可以买到标的)
    stock_prices['result2']=(stock_prices['closePrice']>stock_prices['openPrice']).apply(lambda x:1 if x==True else 0)*2+\
                (stock_prices['nextClosePrice']>stock_prices['closePrice']).apply(lambda x:1 if x==True else 0)
    tmpResult=stock_prices['result2'].value_counts().sort_index()/stock_prices.shape[0]
    print '周一和本周后四天相关性',stockSecID,tmpResult[0]+tmpResult[3]
    
    #print stock_prices[['tradeDate','openPrice','closePrice','nextClosePrice','result1','result2']]

print '### end '

```
