## 顺势指标CCI

- CCI（Commodity Channel lndex）顺势指标是测量股价是否已超出常态分布范围的一个指数，由唐纳德·R.兰伯特（Donald Lambert）所创，属于超买超卖类指标中较特殊的一种，波动于正无限大和负无限小之间。

- 它最早是用于期货市场的判断，后运用于股票市场的研判，并被广泛使用。与大多数单一利用股票的收盘价、开盘价、最高价或最低价而发明出的各种技术分析指标不同，CCI指标是根据统计学原理，引进价格与固定期间的股价平均   区间的偏离程度的概念，强调股价平均绝对偏差在股市技术分析中的重要性，是一种比较独特的技术指标。

- CCI计算过程如下：

- CCI（N日）=（TP－MA）÷ MD ÷ 0.015
  其中，TP =（最高价 + 最低价 + 收盘价）÷ 3
  MA = 近N日收盘价的累计之和 ÷ N
  MD = 近N日（MA－收盘价）的累计之和 ÷ N
  0.015为计算系数，N为计算周期

- 当然我们可以直接用talib来计算CCI，但了解指标的计算方法会帮助我们理解函数中的参数的意义，在使用多个指标的组合时，也避免了使用内核相似的指标来产生重复的信号。

- 本文的研究思路依旧按照如下的五个步骤进行：
  数据准备		
  策略思想
  计算策略收益
  基础回测指标构建
  优化思路

一.数据准备
- 通过DataAPI.MktEqudGet接口调取个股的数据。

```python
import numpy as np
import pandas as pd
import talib as ta
import matplotlib.pyplot as plt

data = DataAPI.MktEqudGet(secID=u"",ticker=u"600030",tradeDate=u"",beginDate=u"20160101",endDate=u"20180101",isOpen="",field=u"",pandas="1")
#调取中信证券的数据
data.set_index('tradeDate', inplace = True)
#将日期设为index

data['cci'] = ta.CCI(np.asarray(data['highestPrice']), np.asarray(data['lowestPrice']), np.asarray(data['closePrice']), timeperiod = 21)
#通过talib求出CCI的值

data['cci'].plot()
plt.title('CCI')
```

二.策略思想
- CCI指标有多种使用方法，其核心思想是：若指标 < -100，说明股票进入超卖区间，即股票被卖的太多，价格偏低，此时应当买入股票；
  若指标 > 100，说明进入超买区间，即股票被买的太多，价格偏高，此时应当卖出股票。

- 我们的思路是，当CCI指标从小于-100的超卖区间进入大于-100的常态区间后，超卖现象消失，股价将趋于上涨，我们做多；
  当CCI指标从大于100的超买区间进入小于100的常态区间后，超买现象消失，股价将趋于下跌，我们做空。

- 'cci_s1'表示将CCI数据向下移动一行，即可代表昨天的数据；
  'cci_s2'表示将CCI数据向下移动两行，即可代表前天的数据；


```python
data['cci_s1'] = data['cci'].shift(1)
data['cci_s2'] = data['cci'].shift(2)
#分别获得代表CCI昨天和前天的数据

data['position'] = np.where((data['cci_s2'] < -100) & (data['cci_s1'] > -100), 1, np.nan)
data['position'] = np.where((data['cci_s2'] > 100) & (data['cci_s1'] < 100), -1, data['position'])
#当出现上文所述的两种情况时，分别产生做多信号（1）和做空信号（-1），其余数据为nan

data['position'] = data['position'].ffill().fillna(0)
#将头寸进行ffill()的向下填充操作，表示持仓情况会一直持续下去
#fillna(0)表示未出现信号时不持仓
```

三.计算策略收益
- pct_change语句可计算每日收益率。
- 将每日收益乘以仓位，即可得到策略收益。
- 注意此处与前两篇文章不同，并未对仓位进行shift(1)的操作，是因为我们是通过昨天和前天的数据得到仓位信号的，这表明在当天是可以通过产生的信号来进行交易的。有疑惑的朋友可以对比前两篇文章产生信号的方式来理解此   处的异同。
- （收益率 + 1）.cumprod方法计算离散的收益率

```python
data['return'] = data['closePrice'].pct_change()
data['strategy'] = data['return'] * data['position']
data['return_cum'] = (data['return'] + 1).cumprod()
data['strategy_cum'] = (data['strategy'] + 1).cumprod()
data[['return_cum', 'strategy_cum']].plot()
```

四.基础回测指标构建
- 构建一个简单的ratio函数来判断收益率、波动率、最大回撤三个指标。

- 注意：输入的数据为每日收益数据，非累积收益数据。

```python
def ratio(data):
    return_year = data.mean() * 252
    #求收益率
    
    volatility = data.std() * 252 ** 0.5
    #求波动率
    
    strategy_cum = (data + 1).cumprod()
    drawdown = (strategy_cum.cummax() - strategy_cum).max()
    #求最大回撤
     
    res = pd.DataFrame([return_year, volatility, drawdown], columns=[data.name], index=['return', 'volatility', 'drawdown'])
    #输出数据为列表[收益率，波动率，最大回撤]， 列名为[data.name]，是取Series下列名的方法， 索引为自定义的['return', 'volatility', 'drawdown']
    return res

ratio(data['return']).join(ratio(data['strategy']))
#DataFrame下的join方法，按index合并，合并benchmark和策略的回测指标
```

五.策略优化
- 通过SMA来判断多空情况，加入CCI作为过滤器。

- 当股价上穿SMA，且CCI < -100时，说明股票是在超卖区间上穿SMA，此时做多信号更为可靠；
  当股价下穿SMA，且CCI > 100时，说明股票是在超买区间下穿SMA，此时做空信号更为可靠；

```python
data['sma'] = ta.SMA(np.asarray(data['closePrice']), 5)
#获取SMA5日均线数据

data['close_s1'] = data['closePrice'].shift(1)
data['close_s2'] = data['closePrice'].shift(2)
#获取收盘价昨天及前天的数据

data['sma_s1'] = data['sma'].shift(1)
data['sma_s2'] = data['sma'].shift(2)
#获取SMA昨天及前天的数据

data['position_sma'] = np.where((data['close_s2'] < data['sma_s2']) & (data['close_s1'] > data['sma_s1']), 1, 0)
data['position_sma'] = np.where((data['close_s2'] > data['sma_s2']) & (data['close_s1'] < data['sma_s1']), -1, data['position_sma'])
#当前天收盘价 < 前天SMA，同时昨天收盘价 > 昨天SMA时，表示股价上穿SMA，仓位为1
#当前天收盘价 > 前天SMA，同时昨天收盘价 < 昨天SMA时，表示股价下穿SMA，仓位为-1

# data['cci_s1'] = data['cci'].shift(1)
#我们之前已经获取了CCI前一天的数据

data['position_cci'] = np.where(data['cci_s1'] < -100, 1, 0)
data['position_cci'] = np.where(data['cci_s1'] > 100, -1, data['position_cci'])
#当昨天CCI < -100时，股票处于超卖区间，仓位为1
#当昨天CCI > 100时，股票处于超买区间，仓位为-1


data['position_mix'] = np.where(data['position_sma'] + data['position_cci'] == 2, 1, np.nan)
data['position_mix'] = np.where(data['position_sma'] + data['position_cci'] == -2, -1, data['position_mix'])
#核心步骤
#当SMA + CCI同时满足做多或做空的信号时，此时再进行做多或做空

data['position_mix'] = data['position_mix'].ffill().fillna(0)
#将头寸进行ffill()的向下填充操作，表示持仓情况会一直持续下去
#fillna(0)表示未出现信号时不持仓

data['strategy_mix'] = data['return'] * data['position_mix']
data['strategy_mix_cum'] = (data['strategy_mix'] + 1).cumprod()
data[['return_cum', 'strategy_cum', 'strategy_mix_cum']].plot()
#画出优化前和优化后的收益图

ratio(data['return']).join(ratio(data['strategy'])).join(ratio(data['strategy_mix']))
#ratio函数来判断收益率、波动率、最大回撤三个指标
```

## 总结
- 只使用CCI的策略效果欠佳。
- SMA + CCI的策略效果不错，不仅收益率大幅提升，从8.09%攀升到了26.15%，而且波动率和最大回撤也都有不小的提升。
- 构建SMA和CCI指标时的时期参数，分别为5和21，这是简单穷举后另结果更优的一组参数，并无其他经济意义。
- 但策略依旧没有考虑做空的限制和交易成本。

## 布林带指标（Bollinger Band）

- BOLL指标是美国股市分析家约翰·布林根据统计学中的标准差原理设计出来的一种非常简单实用的技术分析指标。一般而言，股价的运动总是围绕某一价值中枢（如均线、成本线等）在一定的范围内变动，布林线指标指标正是在上述   条件的基础上，引进了“股价通道”的概念，其认为股价通道的宽窄随着股价波动幅度的大小而变化，而且股价通道又具有变异性，它会随着股价的变化而自动调整。正是由于它具有灵活性、直观性和趋势性的特点，BOLL指标渐渐成   为投资者广为应用的市场上热门指标。

- 布林带（Bollinger Band）是由三条线组成，在中间的通常为20天平均线，而在上下的两条线则分别为Up线和Down线，算法是首先计出过去20日收巿价的标准差SD（Standard Deviation），通常再乘2得出2倍标准差，Up线为   20天平均线加2倍标准差，Down线则为20天平均线减2倍标准差。因为在统计学中，假设正态分布的情况下，上下2倍标准差大概包含了95%以上的情形。
  中间线 = 20日均线
  Up线 = 20日均线 + 2SD
  Down线 =20日均线 - 2SD

一.数据准备
- 使用之前已经获得的个股数据。

```python
data['up'], data['mid'], data['down'] = ta.BBANDS(np.asarray(data['closePrice']), timeperiod=20, nbdevup=2, nbdevdn=2, matype=0) 
#通过talib来计算up线，中线，down线，matype=0表示均线形式是sma
data[['closePrice', 'up', 'mid', 'down']].plot()
```

二.策略思想
- 若股价 < 下轨，说明股票向下偏离的太多，当股价上穿下轨时，认为股价将上涨趋于正常水平，我们做多；
  若股价 > 上轨，说明股票向上偏离的太多，当股价下穿上轨时，认为股价将下跌趋于正常水平，我们做空；

- 'up_s1’表示将布林带上轨的数据向下移动一行，即可代表昨天的数据；
  'up_s2’表示将布林带上轨的数据向下移动两行，即可代表前天的数据；
  同理'down_s1'和'down_s2'表示布林带下轨昨天和前天的数据。

```python
# data['close_s1'] = data['closePrice'].shift(1)
# data['close_s2'] = data['closePrice'].shift(2)
#之前已经获取了收盘价昨天及前天的数据

data['up_s1'] = data['up'].shift(1)
data['up_s2'] = data['up'].shift(2)
#获取布林带上轨昨天及前天的数据

data['down_s1'] = data['down'].shift(1)
data['down_s2'] = data['down'].shift(2)
#获取布林带下轨昨天及前天的数据

data['position_b'] = np.where((data['close_s2'] < data['down_s2']) & (data['close_s1'] > data['down_s1']), 1, np.nan)
#当前天收盘价 < 前天下轨，昨天收盘价 > 昨天下轨时，说明股价上穿下轨，做多

data['position_b'] = np.where((data['close_s2'] > data['up_s2']) & (data['close_s1'] < data['up_s1']), -1, data['position_b'])
#当前天收盘价 > 前天上轨，昨天收盘价 < 昨天上轨时，说明股价下穿上轨，做空

data['position_b'] = data['position_b'].ffill().fillna(0)
#将头寸进行ffill()的向下填充操作，表示持仓情况会一直持续下去
#fillna(0)表示未出现信号时不持仓
```

三.计算策略收益
- pct_change语句可计算每日收益率。
- 将每日收益乘以仓位，即可得到策略收益。
- 注意此处与前两篇文章不同，并未对仓位进行shift(1)的操作，是因为我们是通过昨天和前天的数据得到仓位信号的，这表明在当天是可以通过产生的信号来进行交易的。有疑惑的朋友可以对比前两篇文章产生信号的方式来理解此   处的异同。
- （收益率 + 1）.cumprod方法计算离散的收益率

```python
# data['return'] = data['closePrice'].pct_change()
#之前已经求过

data['strategy_b'] = data['return'] * data['position_b']
data['strategy_b_cum'] = (data['strategy_b'] + 1).cumprod()
data[['return_cum', 'strategy_b_cum']].plot()
```

四.基础回测指标构建
- 构建一个简单的ratio函数来判断收益率、波动率、最大回撤三个指标。

- 注意：输入的数据为每日收益数据，非累积收益数据。

```python
def ratio(data):
    return_year = data.mean() * 252
    #求收益率
    
    volatility = data.std() * 252 ** 0.5
    #求波动率
    
    strategy_cum = (data + 1).cumprod()
    drawdown = (strategy_cum.cummax() - strategy_cum).max()
    #求最大回撤
     
    res = pd.DataFrame([return_year, volatility, drawdown], columns=[data.name], index=['return', 'volatility', 'drawdown'])
    #输出数据为列表[收益率，波动率，最大回撤]， 列名为[data.name]，是取Series下列名的方法， 索引为自定义的['return', 'volatility', 'drawdown']
    return res

ratio(data['return']).join(ratio(data['strategy_b']))
#DataFrame下的join方法，按index合并，合并benchmark和策略的回测指标
```
