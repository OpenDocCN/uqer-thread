# 【大师系列】成长股内在价值投资法——本杰明·格雷厄姆

“价值投资之父”格雷厄姆在其畅销书《聪明的投资者》中给出了一个对成长股内在价值进行估值的简单公式：价值=当期(正常)利润×(8.5 + 两倍的预期年增长率)。并认为该公式就算出的数据，非常接近于一些更加复杂的数学计算所得出的结果。

![image](http://d2.freep.cn/170_3tb_150917135628fr90554431.bmp)

本文中我们用DataAPI.MktStockFactorsOneDayGet中的``'EGRO'``,5年收益增长率平均值来代表预期年增长率，并使用未稀释的``'EPS'``。

根据大师的思想，编写了如下策略，即价格低于公式所表示的价值时买入，价格高于公式所示价值卖出，每日调仓。
