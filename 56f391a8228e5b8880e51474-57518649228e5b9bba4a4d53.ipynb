```python
from hmmlearn.hmm import GaussianHMM
import datetime
import numpy as np
from matplotlib import cm, pyplot as plt
import matplotlib.dates as dates
import pandas as pd
import seaborn as sns
sns.set_style('white')

beginDate = '20120407'
endDate = '20150410'
data = DataAPI.MktIdxdGet(ticker='000001',beginDate=beginDate,endDate=endDate,field=['tradeDate','closeIndex','lowestIndex','highestIndex','turnoverVol'],pandas="1")#获取日行情数据
#print(data) #最前排日期数据，收盘指数，最低指数，最高指数，成交量
data1 = DataAPI.FstTotalGet(exchangeCD=u"XSHE",beginDate=beginDate,endDate=endDate,field=['tradeVal'],pandas="1")#获取沪深融资融券每日汇总信息
#print(data1) #exchangeCD：上海证券交易所 radeVal：本日融资融券余额
data2 = DataAPI.FstTotalGet(exchangeCD=u"XSHG",beginDate=beginDate,endDate=endDate,field=['tradeVal'],pandas="1")
#获取深圳证券交易所的融资融券余额
tradeVal = data1 + data2
#讲两个余额合并
tradeDate = pd.to_datetime(data['tradeDate'][5:])
#print (tradeDate),将第五天日期数据转换 也就是从4月9号开始
volume = data['turnoverVol'][5:]
#volume取上证指数第五天开始的成交量赋值
closeIndex = data['closeIndex']
#closeIndex 保存收盘指数
deltaIndex = np.log(np.array(data['highestIndex'])) - np.log(np.array(data['lowestIndex']))
#先将数据转化成np数组，然后最高最低指数转换将最高最低指数取对数然后相减
deltaIndex = deltaIndex[5:] #取第五天开始的数据
#print(deltaIndex)
logReturn1 = np.array(np.diff(np.log(closeIndex)))
#一日对数收益差,diff 后面的值减前面的值，如果括号里n=2代表相减一次基础上再减一次
logReturn1 = logReturn1[4:]
logReturn5 = np.log(np.array(closeIndex[5:])) - np.log(np.array(closeIndex[:-5]))
#五日对数收益差
logReturnFst = np.array(np.diff(np.log(tradeVal['tradeVal'])))[4:]
#对数融资余额差
closeIndex = closeIndex[5:]
#取第五天开始的收盘指数
X = np.column_stack([logReturn1,logReturn5,deltaIndex,volume,logReturnFst]) #一日对数收益差，五日对数收益差，当日对数高低价差，当日成交量，对数融资余额差

#print(X)
# Make an HMM instance and execute fit
model = GaussianHMM(n_components=6, covariance_type="diag", n_iter=1000).fit([X])#n_components=6模型有6个隐藏的状态covariance_type="diag"需要中等数量的数据n_iter=1000模拟次数=1000次

# Predict the optimal sequence of internal hidden state
hidden_states = model.predict(X)
#print(hidden_states)
#print('6666666')
#print (hidden_states)
#print("Transition matrix")
#print(model.transmat_)
#print()

#print("Means and vars of each hidden state")
#for i in range(model.n_components):
#    print("{0}th hidden state".format(i))
#    print("mean = ", model.means_[i])
#    print("var = ", np.diag(model.covars_[i]))
    
plt.figure(figsize=(15, 8))  #设定画图的尺寸
for i in range(model.n_components):#model.n_components=6代表6个不同的隐藏状态,range(model.n_components)=range(5) =[0,1,2,3,4]一共5个此句意思就是用i遍历6个状态
    #i=0
    idx = (hidden_states==i) #如果hidden_states==i则返回true 否者返回false
    #print(idx)
    #print(tradeDate[idx])
    
    
    
    
    
    
    
    #tradeDate[idx]只留下idx=true的值
    

    plt.plot_date(tradeDate[idx],closeIndex[idx],'.',label='%dth hidden state'%i,lw=1) #plt.plot_date画时间，当hidden_states==i 那天为ture时讲收盘价画上去组成一条线总共有6跳线进行叠加第一个参数时间，后面的Y轴数值，最新日期，收盘指数  画的状态线是'.'点状 %d 第%i种状态，lw=1 线粗=1
    plt.legend()    #显示图例，就是右上角的标签
    plt.grid(1)     #显示网格
    #图1结束
    #整合数据
res = pd.DataFrame({'tradeDate':tradeDate,'logReturn1':logReturn1,'logReturn5':logReturn5,'volume':volume,'state':hidden_states}).set_index('tradeDate')#转化为dataframe数据
#print (res)
plt.figure(figsize=(15, 8)) 
for i in range(model.n_components):
    #print('***********')
    #print(hidden_states)
    idx = (hidden_states==i)
    
    #print(idx)


    #将idx[true false]取第一到最后一个，追加到0后面组成新的idx
    #print('********')
    #print(idx)
    #fast factor backtest
    

    df = res.logReturn1#取res的dataframe中的一日对数收益差赋值给df
    #print(df)

    res['sig_ret%s'%i] = df.multiply(idx,axis=0)#讲df也就是1日指数收益（series或者dataframe格式）差与idx（np.array格式）相乘逐渐生成sig_ret012345 列,此处的i取012345 每个状态 为1的列
    #res['sig_ret%s'%i] =idx
    #print(idx)
    #print(res)
    
    plt.plot(np.exp(res['sig_ret%s'%i].cumsum()),label='%dth hidden state'%i)
    plt.legend()
    plt.grid(1)
    #图2结束
#需要说明的是，这里我对操作做了延时一日的处理，即当天得到某状态后，第二天才能进行操作。可以看到，状态1和状态5是一个明显的上升阶段，可以分别对应于牛市的泡沫期和熊市的反弹期；而状态0是明显的熊市下跌期，状态4是泡沫破裂阶段，状态3更像是某种上升状态后的调整，而状态2就比较subtle，也就是当前的市场状态 :) ，也许可以借鉴一下2010年年末的那段时期？

#简单的策略

#考虑到股指期货可以做空，我们可以设计这样一个策略：当处于状态1和5时第二天做多，当处于状态0、3、4时第二天做空（不管状态2 :) ），其收益曲线如下    
plt.figure(figsize=(15, 8))
idxlong = (hidden_states==1) + (hidden_states==5)  #做多状态
idxlong = np.append(0,idxlong[:-1])#获得状态结果后第二天进行买入操作
idxshort = (hidden_states==0) + (hidden_states==3) + (hidden_states==4) #做空状态
idxshort = np.append(0,idxshort[:-1])#获得状态结果后第二天进行买入操作
#fast factor backtest
df = res.logReturn1
res['sig_retbest'] =  df.multiply(idxlong,axis=0) - df.multiply(idxshort,axis=0)
    
plt.plot_date(tradeDate,np.exp(res['sig_retbest'].cumsum()),'-',label='our strategy')
plt.legend()
plt.grid(1)
#图三结束

#科普 e**2=e^2  loge=ln np.log=loge=ln a=np.exp(np.log(a))   也就是先取a的ln 再exp（lna）=a还原了
```
