```python
import pandas as pd
import numpy as np
# 我将初步处理好的数据保存为zy20160206.xlsx，如果需要，可以留言。
df = pd.read_excel('zy20160206.xlsx','Sheet1',);df.head()     # DataFrame数据写入excel文件
```

## 数据整理
- 将证券代码从 int 转为 str

```python
df.info()
#发现证券代码竟然是 int 。EXCEL中分明是文本！
```

```python
# 如何将 数据处理为 000707 而不是 707 ？
# 步骤：1. 707 + 1000000 1000707 所有都转换
# 2. 将 1000707 转为 str
# 3. 除去第一位str  就可以了
# 还有其他方法：df_secNo.str.len() 计算 如果是7位 ，就str[1:] 可以继续尝试！

df["secID_1"] = df[u"证券代码"].apply(lambda x:str(x+1000000));df.head()
df["secID"] = df["secID_1"].apply(lambda x:x[1:]);df.head()
df = df.drop(["secID_1"],axis=1);df.head()

#自己动手，丰衣足食！
```
