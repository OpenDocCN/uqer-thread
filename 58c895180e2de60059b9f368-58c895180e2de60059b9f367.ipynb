# 价量二维时间序列模式识别与分割

如何给一个股票的历史做阶段性分割包含了对股票历史运动的理解与区分。如果用统计学习的方法来做，就是对一个二维变量的时间序列进行time series segmentation.有兴趣的同学可以谷歌这三个英文关键词帮助你深入研究。

受J.Abonyi,B.Feil,S.Nemeth,and P.Avra 2004年的文章 Principal component analysis based time series segmentation: A new sensor fusion algorithm启发，我想通过识别股票价格与成交量的correlation变化来分割股票历史阶段。用到的方法，通俗来讲是用Singular-Value-Decomposition的right sigular vector投影原始数据，然后通过计算投影与原始数据的error，判断原始数据的变量之间是否出现大的correlation变动。

这个分割技术有什么用呢？我觉得它是momentum方法的帮手，所谓高买高卖，低买低卖是也。详见下文。

```python
import datetime
import math
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
from CAL.PyCAL import *
cal = Calendar('China.SSE')
font.set_size(15)
todaydate=datetime.datetime.now().strftime('%Y-%m-%d')
lastTD=cal.advanceDate(todaydate,'-1B',BizDayConvention.Preceding).strftime('%Y-%m-%d')#last trading date
datah=DataAPI.MktEqudGet(secID=u"",ticker='000001'
                      ,isOpen=u"1",beginDate=u'1999-06-28',endDate=todaydate
                                ,field=u"secShortName,tradeDate,closePrice,turnoverVol,accumAdjFactor,chgPct"
                              ,pandas="1")
datah.head()
```

```python
##复权处理
datah['closePrice']=datah['closePrice']*datah['accumAdjFactor']
datah['turnoverVol']=datah['turnoverVol']/datah['accumAdjFactor']
datah.head()
```

分割的思想采用了一种bottom-up的算法，即先在一开始做一个很密的分割，然后通过不断merge相邻的segment来达到最后的合理分割结果。
对于最开始的密集分割，很自然的，我选择了local extrema。通过改动差分的阶数order可以控制密集程度。

```python
##initial critical points--local extrema
from scipy.signal import argrelextrema
locmax =  argrelextrema(datah['closePrice'].values,
                     comparator=np.greater,
                    order=20)
plt.figure(figsize=(18,6))
plt.plot(datah['closePrice'].values,'^',ls='-',markevery=list(locmax)[0])

locmin =argrelextrema(datah['closePrice'].values, np.less,order=20)#return tuple
plt.figure(figsize=(18,6))
plt.plot(datah['closePrice'].values,'v',ls='-',markevery=list(locmin)[0])
```

计算SVD投影的error，由Q-measure计算，详情可查。
对当前的分割，如果某两个相邻的segment合并后的error是最小的，那么久合并该两个segment，并更新分割，直至达到停止准则。

```python
#initial fine-grain
loc=np.concatenate((list(locmax)[0],list(locmin)[0]),axis=0)
loc.sort()
```

```python
from sklearn import preprocessing
def mergecost(x,i,loc,p=1):
    if i==0:
        s1=x.iloc[:(loc[i+1]+1)]
    elif i<(len(loc)-1):
        s1=x.iloc[(loc[i-1]+1):(loc[i+1]+1)]  
    else:
        s1=x.iloc[(loc[i-1]+1):]  
        
    a=s1[['closePrice','turnoverVol']]
    a_scaled = preprocessing.scale(a)##scale and center
    U, s, V = np.linalg.svd(a_scaled, full_matrices=True)#SVD of 1st segment
    S = np.zeros(a.shape, dtype=complex)
    S[:a.shape[1], :a.shape[1]] = np.diag(s)
    p=p#projection dimension
    pro=np.dot(a_scaled,np.transpose(V[0:p,]))
    p=np.dot(pro,V[0:p,])#reconstruction by dimension 1
    #Q1=np.mean(np.square(p))
    error=a_scaled-p
    Q2=np.mean(np.diagonal(np.dot(error,np.transpose(error))))
    return Q2    

def bottomup(dat,seg_num,initial_loc,p=1):
    loc = initial_loc
    mc=[mergecost(dat,i,loc=loc,p=p) for i in range(len(loc))]
    while len(loc) > seg_num:
        j=mc.index(min(mc))#seg to merge
        loc=np.delete(loc,j)#update loc
        mc=np.ndarray.tolist(np.delete(mc,j))
        
        if j==0:#update mc
            mc[0]=mergecost(dat,0,loc=loc,p=p)
        elif j<len(loc):    
            mc[j-1]=mergecost(dat,j-1,loc=loc,p=p)
            mc[j]=mergecost(dat,j,loc=loc,p=p)
        else:
            mc[j-1]=mergecost(dat,j-1,loc=loc,p=p)
    return loc
```

```python
loc_final=bottomup(datah,50,loc,1)
```

```python
plt.figure(figsize=(18,6))
plt.plot(datah['closePrice'].values,'*',ls='-',markevery=loc_final)
plt.figure(figsize=(18,6))
plt.bar(range(len(datah['turnoverVol'].values)),datah['turnoverVol'].values)
```

五角星代表最后的分割点。有意思的是，在这个分割准则下，如果单看价格，很有可能发现在同一个阶段内价格经历了一段下跌趋势与上涨趋势。比如图中1041~1847（点的序列，对应时间搜索datah即可）。

但如果清楚该分割本质的同学会明白，虽然1041~1847出现了价格数据的趋势变动，但价格与成交量两者的相关关系并没有变化，价跌量也跌，故该segment成立。
