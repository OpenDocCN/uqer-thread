```python
# 大盘周数据
# DataAPI.MktIdxwGet(beginDate=u"20160101",endDate=u"20160812",indexID=u"000001.ZICN",ticker=u"",field=u"",pandas="1")
# 大盘日数据
# DataAPI.MktIdxdGet(tradeDate=u"",indexID=u"000001.ZICN",ticker=u"",beginDate=u"20160101",endDate=u"20160110",field=u"",pandas="1")
# 计算30周移
universe = ['000001.XSHE', '600000.XSHG']
from CAL.PyCAL import *
import pandas as pd
import matplotlib.pylab as plt
import numpy as np
def line_mean(stock,feq='30W',length=20,tradeDay='2014-01-23'):
    #数据容器：字典
    result={}.fromkeys(stock)
    #对每一个进行循环
    for stk in stock:
        #初始化
        beginDate=Date.parseISO(tradeDay)-Period(feq)
        endDate=Date.parseISO(tradeDay)+Period(str(length)+'D')
        line_data=[]
        #对长度循环
        data=DataAPI.MktEqudGet(tradeDate=u"",secID=stk,ticker=u"",beginDate=beginDate.toISO(),endDate=endDate.toISO(),isOpen=u"",field=u"secID,closePrice,tradeDate",pandas="1")
        endDate=Date.parseISO(tradeDay)
        for l in range(length):
            date=pd.date_range(start=beginDate.toISO(),end=endDate.toISO(),freq='W-FRI').strftime('%Y-%m-%d')           
            data['if_1']=np.nan
            for i in range(len(data)):
                if data.tradeDate[i] in date: 
                    data.if_1[i]=1
            data_temp=data.dropna()           
            temp=data_temp[data_temp['secID']==stk].mean()['closePrice']
            line_data.append(temp)
            beginDate=beginDate+Period('1D')
            endDate=endDate+Period('1D')
            result[stk]=line_data
    return result
data1=line_mean(stock=universe) 
plt.plot(data1['000001.XSHE'])
```
