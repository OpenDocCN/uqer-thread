# 根据六大特征寻找潜在壳资源
---
 
- 1、筛选全A中市值最小的前500名（剔除不能借壳上市的创业板企业）；
- 2、剔除主要股东持股比例过低的股票；
- 3、剔除净利润增速大于0的股票；
- 4、剔除ROE大于0.05的上市公司；
- 5、剔除所属行业壳资源公司占比较少的公司（银行、军工、非银行金融、传媒等）；
- 6、按照EPS等基本面指标排序打分；

```python
import pandas as pd
import numpy as np
```

# 1. 市值最小500只

```python
secs = list(set(set_universe('A'))-set(set_universe('CYB')))  # 去除创业板的全A股股票池
data1 = DataAPI.MktStockFactorsOneDayGet(secID=secs,tradeDate='20160315',field=['secID','ticker','tradeDate','LCAP'])
data1[u'市值/亿'] = [np.floor(np.exp(x-np.log(1e8))) for x in data1.LCAP]
data1 = data1.sort('LCAP').head(500)
equ_info = DataAPI.EquGet(equTypeCD=u"A",secID=secs,field=u"secID,partyID,secShortName")
data1 = pd.merge(equ_info, data1, on='secID')
data1 = data1.sort('LCAP')
data1.tail()
```

# 2. 剔除主要股东持股比例过低的股票

```python
data2_tmp = data1.set_index('partyID')
data2_tmp[u'大股东持比'] = 0.0
data2_tmp[u'十大股东持比'] = 0.0
for party in data2_tmp.index:
    # 获取十大股东信息
    fields = u"partyID,endDate,shRank,holdPct"
    tmp = DataAPI.JY.EquMainshJYGet(partyID=party,infoTypeCD=u"1",endDateStart=u"20150601",endDateEnd=u"",field=fields)
    tmp = tmp[tmp.endDate==tmp.endDate.values[0]]
    data2_tmp[u'大股东持比'][party] = np.floor(tmp[tmp.shRank==1]['holdPct'].values[0])
    data2_tmp[u'十大股东持比'][party] = np.floor(tmp['holdPct'].values.sum())

data2_tmp = data2_tmp.reset_index()
data2_tmp = data2_tmp.sort('LCAP')
data2_tmp.head()
```

```python
# 按持股比例筛选
data2 = data2_tmp[data2_tmp[u'大股东持比'] > 10.0][data2_tmp[u'十大股东持比'] > 20.0]
data2 = data2.sort('LCAP')
print len(data2)
data2.head()
```

# 3. 剔除净利润增长大于0的股票

```python
tmp = DataAPI.MktStockFactorsOneDayGet(secID=data2.secID,tradeDate='20160315',field=['secID','NPParentCompanyGrowRate'])
tmp.columns = ['secID', u'净利润增长率']
data3 = pd.merge(data2, tmp, on='secID')
data3 = data3[~(data3[u'净利润增长率'] > 0)]
data3 = data3.sort('LCAP')
print len(data3)
data3.head(5)
```

# 4. 剔除ROE大于0.05的上市公司

```python
tmp = DataAPI.MktStockFactorsOneDayGet(secID=data3.secID,tradeDate='20160315',field=['secID','ROE'])
data4 = pd.merge(data3, tmp, on='secID')
data4 = data4[~(data4.ROE > 0.05)]
data4 = data4.sort('LCAP')
print len(data4)
data4.head(5)
```

# 5. 剔除特殊行业

```python
tmp = DataAPI.EquIndustryGet(industryVersionCD=u"010303",secID=data4.secID,field=u"secID,industryName1")
tmp = tmp.groupby('secID').first().reset_index()
data5 = pd.merge(data4, tmp, on='secID')
data5 = data5[~(data5.industryName1 == '国防军工')]
data5 = data5[~(data5.industryName1 == '传媒')]
data5 = data5[~(data5.industryName1 == '汽车')]
data5 = data5[~(data5.industryName1 == '农林牧渔')]
data5 = data5[~(data5.industryName1 == '房地产')]
data5 = data5[~(data5.industryName1 == '电子')]
data5 = data5[~(data5.industryName1 == '公用事业')]
data5 = data5[~(data5.industryName1 == '钢铁')]

data5 = data5.sort('LCAP')
print len(data5)
data5.head(5)
```

# 6. 按基本面因子打分排序

```python
# 按照下面五个因子排序
fields = ['secID',
          'EPS',  # 每股收益
          'CTOP', # 现金流市值比
          'PCF',  # 市现率
         ]
tmp = DataAPI.MktStockFactorsOneDayGet(secID=data5.secID,tradeDate='20160315',field=fields)

# 市现率实际中一般越小越好，此处取将市现率去倒数，方便排序比较！！！
tmp['PCF'] = [1/x for x in tmp.PCF]
data6 = pd.merge(data5, tmp, on='secID')

data6 = data6.fillna(-100) # 为了排序方便，将NAN补为-100

# 打分排序
data6[u'壳排序'] = data6.EPS.rank() + data6[u'大股东持比'].rank(ascending=False) + data6.CTOP.rank() + data6.PCF.rank() + data6.ROE.rank()
data6[u'壳排序'] = data6[u'壳排序'].rank()
data6 = data6.set_index(u'壳排序').sort_index()

cols = [x for x in data6.columns.values if x not in ['ticker','LCAP','tradeDate','partyID']]
data6 = data6[cols]

# 去除前面XX名后按市值重新排序
data6.head(60).sort(u'市值/亿')
```
