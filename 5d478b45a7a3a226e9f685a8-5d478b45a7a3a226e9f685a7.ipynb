下面是**资产翻倍期**在帖子中的定义：

> * 评价指标很多有市盈率、市净率、资产收益率等等。如何综合量化是个难题。这里我提出我“资产翻倍期”的观点。即假定公司不分红，资产收益率不变，每年的收益利滚利，达到股价两倍所需要的时间。 在EXCEL表中用函数:LOG(市净率两倍，资产收益率+1)，翻倍期越小越好。
