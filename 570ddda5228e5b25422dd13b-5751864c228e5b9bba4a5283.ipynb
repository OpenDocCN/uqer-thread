第一种方法：先一次性取出交易时段所需数据

```python
def FactorGet(universe, factor_days,window):
    all_data = {} #record the return data; key:stock_id; value: factor value in trading_days (in Pandas)
    #数据窗口期前移
    delta = '-'+str(window)+'B'
    data_start = cal.advanceDate(factor_days[0],delta,BizDayConvention.Preceding)    
    for ticker in universe:
        all_data[ticker] = DataAPI.MktEqudGet(secID = ticker, beginDate = data_start.strftime('%Y%m%d'), endDate = factor_days[-1], field =['tradeDate','closePrice','openPrice','highestPrice','lowestPrice'],pandas="1")
    return all_data
```

以上代码取出了universe中包含的股票在trading_days这一时间段内所有的数据，并返回key为stock ID， value为时间段内所有因子数据的DataFrame。之所以这样一次性取出整个时间段的数据，是利用了MktEqudGet可以取单只股票连续时间段内的因子数据，减少在handle_data中逐天调用DataAPI带来的延迟。在实际回测使用中，我们需要的一天中所有股票的数据，利用以下函数对产生信号

```python
#Quartz Signal
#收益波动比因子，rwR = (closePrice-openPrice)/(highPrice-lowPrice),式中：open为开盘买入价格，close为卖出价格，high为持有期内最高价，low为持有期内最低价。

def rwR(univ,lastTDay,window):
    pd_dict = {}
    day_str = lastTDay.strftime('%Y-%m-%d')
    stock_pd = DataFrame()
    for stock in univ:
        stock_pd = factor_data[stock]
        stock_pd = stock_pd[(stock_pd.tradeDate<=day_str)].sort(columns=['tradeDate'],ascending=False).head(window)
        #因子处理逻辑
        clsps = stock_pd['closePrice'].values
        opeps = stock_pd['openPrice'].values
        higps = stock_pd['highestPrice'].values
        lowps = stock_pd['lowestPrice'].values

        high = max(higps[-6:-1])
        low  = min(lowps[-6:-1])

        pd_dict[stock] = (clsps[-1]-opeps[-5])*1.0/(high - low+0.0001)
   
    #引入 Series 
    ret_dict  = pd.Series(pd_dict)
    ret_dict.sort(ascending = False)
    ret_dict.dropna(inplace = True)
    return ret_dict
```

```python
#筛除不合格的股票
def ini_univ(account,lastTDay):
    univ = account.universe
    # 1. 去除停牌股票
    univ = [s for s in univ if s in account.universe]

    # 2. 去除ST股
    try:
        STlist = DataAPI.SecSTGet(secID=univ, beginDate=lastTDay, endDate=lastTDay, field=['secID']).tolist()
        univ = [s for s in univ if s not in STlist]
    except:
        pass

    # 3. 去除流动性差的股票
    tv = account.get_attribute_history('turnoverValue', 20)
    mtv = {sec: sum(tvs)/20. for sec, tvs in tv.iteritems()}
    univ = [s for s in univ if mtv.get(s, 0) >= 10**7]

    # 4. 去除新上市或复牌的股票
    opn = account.get_attribute_history('openPrice', 1)
    univ = [s for s in univ if not (np.isnan(opn.get(s, 0)[0]) or opn.get(s, 0)[0] == 0)]
    return univ
```

```python
import numpy as np

from pandas import DataFrame,Series
import pandas as pd

import quartz as qz
from quartz.api import *
from CAL.PyCAL import *
cal = Calendar('China.SSE')

from datetime import datetime, timedelta

import datetime
starttime = datetime.datetime.now()


# -----------回测参数部分开始，可编辑------------
#回测区间
start = '2012-08-01'                       # 回测起始时间
end   = '2015-08-01'                         # 回测结束时间

#投资域：自定，但需去除ST股；去除流动性差的股票；去除新上市或复牌的股票；剔除了当天停牌、退市和数据异常证券。
#universe = set_universe('SH50')            # 上证50成份股
#print list(IndSW.L1()) #申万一级行业
universe = set_universe('HS300')           # 沪深300成份股
#universe = set_universe('ZZ500')           # 中证500成份股


#策略基准
benchmark = 'HS300'                          # 策略参考标准

#账户基准
capital_base = 1000000                    # 起始资金
security_base = {}                        # 初始证券头寸
commission = Commission(buycost=0.0008, sellcost=0.0018)  # 佣金万八
slippage = Slippage()

freq = 'd'                                 # 策略类型，'d'表示日间策略使用日线回测，'m'表示日内策略使用分钟线回测
refresh_rate = 20                         # 调仓频率，表示执行handle_data的时间间隔，若freq = 'd'时间间隔的单位为交易日，若freq = 'm'时间间隔为分钟

window = 20   # 取行情数据窗口期
 
factor_days = qz.utils.tradingcalendar.get_trading_days(start, end)

factor_data = FactorGet(universe, factor_days,window) # 取因子或行情数据,考虑窗口期

def initialize(account):                   # 初始化虚拟账户状态
    pass
def handle_data(account):                  # 每个交易日的买入卖出指令          # 包含stk 和benchmark

    #获取前一交易日
    lastTDay = cal.advanceDate(account.current_date,'-1B',BizDayConvention.Preceding)  # 获取上一个交易日

    # 合规处理
    univ = ini_univ(account,lastTDay)

    #自定义信号
    sig = rwR(univ,lastTDay,window)
    #信号处理
    sig = standardize(neutralize(winsorize(sig), lastTDay.strftime('%Y%m%d')))

    # 获得权重
    wts = simple_long_only(sig,lastTDay.strftime('%Y%m%d')) 

    # 卖掉不在权重字典当中的股票
    for stock in account.valid_secpos:
        if stock in account.universe and stock not in wts:
            order_to(stock, 0)

    # 根据权重计算调仓数量
    c = account.referencePortfolioValue
    change = {}
    for stock, w in wts.iteritems():
        p = account.referencePrice[stock]
        if p and not np.isnan(p):
            change[stock] = int(c * w / p) - account.valid_secpos.get(stock, 0)

    # 按先卖后买的顺序发出指令
    for stock in sorted(change, key=change.get):
        order(stock, change[stock])
    return

sim_params = qz.sim_condition.env.SimulationParameters(start, end, benchmark, universe, capital_base)
idxmap_all, data_all = qz.sim_condition.data_generator.get_daily_data(sim_params)

strategy = qz.sim_condition.strategy.TradingStrategy(initialize, handle_data)        
bt, acct = qz.quick_backtest(sim_params, strategy, idxmap_all, data_all, refresh_rate = refresh_rate)
perf = qz.perf_parse(bt, acct)
#画图
perf['cumulative_returns'].plot()
#计时
endtime = datetime.datetime.now()
interval=(endtime - starttime).seconds
print '此程序共运行 %s（秒）' % interval
```

第二种方法：按#Quartz Signal生成数据

```python
#Quartz Signal
#收益波动比因子，rwR = (closePrice-openPrice)/(highPrice-lowPrice),式中：open为开盘买入价格，close为卖出价格，high为持有期内最高价，low为持有期内最低价。

def rwR(data, dependencies=['closePrice','openPrice', 'highPrice','lowPrice'], max_window=20):
    temp_result = {}
    for ticker in data['openPrice'].columns:
        clsps = data['closePrice'][ticker].values
        opeps = data['openPrice'][ticker].values
        higps = data['highPrice'][ticker].values
        lowps = data['lowPrice'][ticker].values

        high = max(higps[-6:-1])
        low  = min(lowps[-6:-1])

        temp_result[ticker] = (clsps[-1]-opeps[-5])*1.0/(high - low+0.0001)
    ret = pd.Series(temp_result)
    ret.sort(ascending = False)
    ret.dropna(inplace = True)
    return ret
```

```python
import numpy as np

from pandas import DataFrame,Series
import pandas as pd

import quartz as qz
from quartz.api import *
from CAL.PyCAL import *
cal = Calendar('China.SSE')

from datetime import datetime, timedelta

import datetime
starttime = datetime.datetime.now()


# -----------回测参数部分开始，可编辑------------
#回测区间
start = '2012-08-01'                       # 回测起始时间
end   = '2015-08-01'                         # 回测结束时间

#投资域：自定，但需去除ST股；去除流动性差的股票；去除新上市或复牌的股票；剔除了当天停牌、退市和数据异常证券。
#universe = set_universe('SH50')            # 上证50成份股
#print list(IndSW.L1()) #申万一级行业
universe = set_universe('HS300')           # 沪深300成份股
#universe = set_universe('ZZ500')           # 中证500成份股


#策略基准
benchmark = 'HS300'                          # 策略参考标准

#账户基准
capital_base = 1000000                    # 起始资金
security_base = {}                        # 初始证券头寸
commission = Commission(buycost=0.0008, sellcost=0.0018)  # 佣金万八
slippage = Slippage()

freq = 'd'                                 # 策略类型，'d'表示日间策略使用日线回测，'m'表示日内策略使用分钟线回测
refresh_rate = 20                         # 调仓频率，表示执行handle_data的时间间隔，若freq = 'd'时间间隔的单位为交易日，若freq = 'm'时间间隔为分钟


def initialize(account):                   # 初始化虚拟账户状态
    a = Signal("mysig", rwR)
    account.signal_generator = SignalGenerator(a)

def handle_data(account):                  # 每个交易日的买入卖出指令          # 包含stk 和benchmark

    #获取前一交易日
    lastTDay = cal.advanceDate(account.current_date,'-1B',BizDayConvention.Preceding)  # 获取上一个交易日

  
    # 合规处理
    univ = ini_univ(account,lastTDay)

    #自定义信号
    sig = account.signal_result['mysig']
    
    #信号处理
    sig = {s: sig[s] for s in univ}
    sig = standardize(neutralize(winsorize(sig), lastTDay.strftime('%Y%m%d')))

    # 获得权重
    wts = simple_long_only(sig,lastTDay.strftime('%Y%m%d')) 

    # 卖掉不在权重字典当中的股票
    for stock in account.valid_secpos:
        if stock in account.universe and stock not in wts:
            order_to(stock, 0)

    # 根据权重计算调仓数量
    c = account.referencePortfolioValue
    change = {}
    for stock, w in wts.iteritems():
        p = account.referencePrice[stock]
        if p and not np.isnan(p):
            change[stock] = int(c * w / p) - account.valid_secpos.get(stock, 0)

    # 按先卖后买的顺序发出指令
    for stock in sorted(change, key=change.get):
        order(stock, change[stock])
    return

sim_params = qz.sim_condition.env.SimulationParameters(start, end, benchmark, universe, capital_base)
idxmap_all, data_all = qz.sim_condition.data_generator.get_daily_data(sim_params)

strategy = qz.sim_condition.strategy.TradingStrategy(initialize, handle_data)        
bt, acct = qz.quick_backtest(sim_params, strategy, idxmap_all, data_all, refresh_rate = refresh_rate)
perf = qz.perf_parse(bt, acct)
#画图
perf['cumulative_returns'].plot()
#计时
endtime = datetime.datetime.now()
interval=(endtime - starttime).seconds
print '此程序共运行 %s（秒）' % interval
```
