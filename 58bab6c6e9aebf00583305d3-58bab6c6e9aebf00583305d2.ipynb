# 2017-0304-画个股票蜡烛图-发布代码

```python


import DataAPI
import numpy as np
import pandas as pd
import matplotlib as mat
import matplotlib.pyplot as plt



start = '2016-11-01'
end = '2016-12-20'
ticker = '000757'
df = DataAPI.MktEqudAdjGet(ticker = ticker, beginDate = start, endDate = end, field = ['tradeDate', 'openPrice', 'closePrice', 'highestPrice', 'lowestPrice'])
print df

##上涨的
shangzhang = df.copy()
shangzhang['closePrice'][df.closePrice < df.openPrice] = np.nan ##注意这里将开盘价等于收盘价的视为红色
shangzhang.ix[df.closePrice < df.openPrice, 'openPrice'] = np.nan
tag = np.array(range(0, len(df)))
xzuo = tag - 0.2
xyou = tag + 0.2

fig = plt.figure(figsize = (16, 8))
ax = fig.add_subplot(1, 1, 1)

for i in df.index:
    if shangzhang.openPrice[i] is np.nan:  
        continue;
    zuoxia = [xzuo[i], shangzhang.openPrice[i]]
    youxia = [xyou[i], shangzhang.openPrice[i]]
    youshang = [xyou[i], shangzhang.closePrice[i]]
    zuoshang = [xzuo[i], shangzhang.closePrice[i]]
    rectred = mat.patches.Polygon([zuoxia, youxia, youshang, zuoshang], color = 'r')
    ax.add_patch(rectred)
    linesh = mat.lines.Line2D([tag[i], tag[i]], [shangzhang.highestPrice[i], shangzhang.closePrice[i]], color = 'r')
    ax.add_line(linesh)
    linexia = mat.lines.Line2D([tag[i], tag[i]], [shangzhang.openPrice[i], shangzhang.lowestPrice[i]], color = 'r')
    ax.add_line(linexia)
    
plt.axis([-2, len(df), df.lowestPrice.min() - 1.0, df.highestPrice.max() + 1.0])  
shangzhang['highestPrice'][df.closePrice < df.openPrice] = np.nan
shangzhang.ix[df.closePrice < df.openPrice, 'lowestPrice'] = np.nan

##下跌
xiadie = df.copy()
xiadie['closePrice'][df.closePrice >= df.openPrice] = np.nan ##
xiadie.ix[df.closePrice >= df.openPrice, 'openPrice'] = np.nan
xiadie['highestPrice'][df.closePrice >= df.openPrice] = np.nan  ##两种能赋值的, 都写一下, 就当是练一下
xiadie.ix[df.closePrice >= df.openPrice, 'lowestPrice'] = np.nan

for i in df.index:
    if xiadie.openPrice[i] is np.NAN:
        continue;
    zuoxia = [xzuo[i], xiadie.openPrice[i]]
    youxia = [xyou[i], xiadie.openPrice[i]]
    youshang = [xyou[i], xiadie.closePrice[i]]
    zuoshang = [xzuo[i], xiadie.closePrice[i]]
    rectred = mat.patches.Polygon([zuoxia, youxia, youshang, zuoshang], color = 'g')
    ax.add_patch(rectred)
    linesh = mat.lines.Line2D([tag[i], tag[i]], [xiadie.openPrice[i], xiadie.highestPrice[i]], color = 'g')
    ax.add_line(linesh)
    linexia = mat.lines.Line2D([tag[i], tag[i]], [xiadie.closePrice[i], xiadie.lowestPrice[i]], color = 'g')
    ax.add_line(linexia)
    
    
ax.set_xticklabels(df.tradeDate)
ax.set_xticks(np.linspace(0,len(df)-1,len(df))) 
plt.setp(plt.gca().get_xticklabels(),rotation=90,horizontalalignment='center')
plt.show()    
```
