# 【 大师系列 】成功选股法 —— 柏顿.墨基尔

##  柏顿.墨基尔 成功选股法

* `柏顿·墨基尔(Burton G. Malkiel)`华尔街专业投资人、经济学者。


>*  柏顿·墨基尔,著名的美国投资大师,现任普林斯顿大学(Princeton University)客座教授，曾任职于史密斯巴尼(Smith Barney & Co.)投资银行部门及数家大型投资机构的董事，如先锋集团(Vanguard Group of Investment Companies)，保德信人寿(Prudential InsuranceCompany of America)等，也曾获聘为美国总统经济谘询委员会的委员，还身兼美国普林斯顿华友银行总裁、美国一家投资额达2000 亿美元的保险公司的财务委员，在学术界及投资界，均是重量级的大师级人物。


>*  知名著作 ：

>  《漫步华尔街》(A Random Walk down
Wall Street- Including A life-cycle Guide to Personal Investing)


> > `选股条件` ：


>     1.   最近四季度市盈率最低的30%；(PE(TTM))  

>     2.   最近四季度净利润(TTM)增长最高的30% 

>     3.   连续5年内盈余成长率为正值；（最近3年净利润未出现亏损 11年之前）
