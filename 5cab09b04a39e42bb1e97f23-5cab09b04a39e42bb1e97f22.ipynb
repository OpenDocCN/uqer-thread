# 4.策略分析
### - 这部分要做的事包括：
- 1、此部分要将数据按照“增持比例”进行分组
- 2、验证各增持比例中，在增持前60天至增持后120天当中股票的表现，是否能获得收益？能不能获得超额收益？
### - 验证说明：
- 将所有的A股按照规模分为20组，计算每个分组的平均收益率，以此作为基准收益率
- 将个股收益率减去个股所在分组的收益率，计算得到超额收益率
### - 窗口期说明：
- 设置窗口期为[-60,120]，原因在于：
	- 观察个股在增持前的股价情况，预计为负，股价被低估，所以高管进行增持行为
	- 观察在增持前几天（这里设置3天）的股价情况，若股价大多出现上升，那么是否能说明存在增持消息提前泄露的情况？
	- 观察在增持公告发布后的股价表现，以此为基础设置策略中的持仓期限

```python
"""
    定义几个函数：
        1.给定日期，可以通过这个函数获取前后N天的那个交易日
        2.获取个股的一个时间段的收益
        3.获取一个股票对应的市值组合的一段时间的收益
        4.获取个股的超额收益
        5.获取各个增持比例下[-60,120]的相对收益率
    
"""

def get_specific_trading_day(date_str='2016-01-04', window=20, direction='after'):
    """给定日期，可以通过这个函数获取前后N天的那个交易日"""
    
    # 假如日期格式为str类型，需要转换成日期格式
    date = datetime.strptime(date_str, '%Y-%m-%d')
    # 假如日期格式已经是datetime类型，就不用进行转换了，直接用下面这个
    # date = date_str
    
    if direction=='before':
        begin_day = date - (window+5) * 4 * timedelta(days=1)   # 可以获取冗余的交易日，防止最终取不到有效数据
        end_day = date
    elif direction=='after':
        begin_day = date
        end_day = date + (window+5)  * 4 * timedelta(days=1) 
        
    begin_day_str = begin_day.strftime('%Y-%m-%d')
    end_day_str = end_day.strftime('%Y-%m-%d')
    
    cal_dates = DataAPI.TradeCalGet(exchangeCD=u"XSHG", beginDate=begin_day_str, endDate=end_day_str, field="calendarDate,isOpen")
    trading_days = cal_dates[cal_dates['isOpen']==1]["calendarDate"].tolist()

    try:
        if direction=='before':
            return trading_days[-window-1]
        return trading_days[window]
    except:
        print date_str,window,direction


def get_ret(ticker, begin_day, end_day):
    """获取个股的一个时间段的收益"""
    begin_close = DataAPI.MktEqudAdjGet(tradeDate=begin_day,secID=u"",ticker=ticker,isOpen="",beginDate=u"",endDate=u"",field=u"closePrice",pandas="1")['closePrice']

    end_close = DataAPI.MktEqudAdjGet(tradeDate=end_day,secID=u"",ticker=ticker,isOpen="",beginDate=u"",endDate=u"",field=u"closePrice",pandas="1")['closePrice']
    try:
        ret = end_close[0] / begin_close[0] -1.0
        return ret
    # 可能存在停牌等情况
    except:
        print 'get_ret error: ' +ticker+begin_day+end_day
        return 999999999


def get_base_ret(ticker='000651', begin_day='2007-09-26', end_day='2007-12-19'):
    """获取一个股票对应的市值组合的一段时间的收益"""
    mkt = DataAPI.MktEqudGet(tradeDate=begin_day,secID=u"",ticker=u"",beginDate=u"",endDate=u"",isOpen="",field=u"tradeDate,ticker,marketValue",pandas="1")
    # 分组
    mkt['label'] = pd.qcut(mkt['marketValue'], 20, labels=False)
    label = mkt[mkt['ticker']=='000001']['label'][0]
    base_df = mkt[mkt['label']==label]
    base_tickers = list(base_df['ticker'])

    begin_close = DataAPI.MktEqudAdjGet(tradeDate=begin_day,secID=u"",ticker=base_tickers,isOpen="",beginDate=u"",endDate=u"",field=u"ticker,closePrice",pandas="1")
    
    end_close = DataAPI.MktEqudAdjGet(tradeDate=end_day,secID=u"",ticker=base_tickers,isOpen="",beginDate=u"",endDate=u"",field=u"ticker,closePrice",pandas="1")
    
    result_df = begin_close.merge(end_close, how='inner',on='ticker')
    result_df['ret'] = result_df['closePrice_y'] / result_df['closePrice_x'] -1.0
    
    return result_df['ret'].mean()


def get_alpha(level_data,begin='60_days_ago',end='4_days_ago'):
    """获取个股的超额收益"""
    ret_dict = {}
    # 分年统计
    for year in range(2011, 2019):
        print year
        year_df = level_data[level_data['year']==year]
        alpha_ls = []
        # 遍历一下
        for row in year_df.iterrows():
            ticker= row[1]['ticker']
            begin_day = row[1][begin]
            end_day = row[1][end]
            single_ret = get_ret(ticker, begin_day, end_day)
            # 以continue跳出循环的方式剔除异常值
            if single_ret > 99999:
                continue
            bese_ret = get_base_ret(ticker, begin_day, end_day)
            alpha_ls.append(single_ret - bese_ret)
        ret_dict[str(year)]=np.mean(alpha_ls)

    return pd.Series(ret_dict)


def get_alpha_df(level_data):
    """获取各个增持比例下[-60,120]的相对收益率"""
    alpha_df = pd.DataFrame()
    alpha_df['before_60_to_today'] = get_alpha(level_data,'60_days_ago','date')
    alpha_df['before_3_to_today'] = get_alpha(level_data,'3_days_ago','date')
    alpha_df['today_to_after_1'] = get_alpha(level_data,'date','1_days_after')
    alpha_df['today_to_after_3'] = get_alpha(level_data,'date','3_days_after')
    alpha_df['today_to_after_5'] = get_alpha(level_data,'date','5_days_after')
    alpha_df['today_to_after_10'] = get_alpha(level_data,'date','10_days_after')
    alpha_df['today_to_after_20'] = get_alpha(level_data,'date','20_days_after')
    alpha_df['today_to_after_30'] = get_alpha(level_data,'date','30_days_after')
    alpha_df['today_to_after_40'] = get_alpha(level_data,'date','40_days_after')
    alpha_df['today_to_after_60'] = get_alpha(level_data,'date','60_days_after')
    alpha_df['today_to_after_90'] = get_alpha(level_data,'date','90_days_after')
    alpha_df['today_to_after_120'] = get_alpha(level_data,'date','120_days_after')
    
    return alpha_df
```

# 几个观察结果：
- 1、2015年的增持效果最好，2017年的增持效果最差
	- 其中原因可能在于2015年上半年的牛市行情中，增持个股所获收益将相对收益拉高，而在年末的急跌中基准和个股的相对收益并不明显；对于2017年的表现，观察发现2017年的增持公告主要集中在下半年时期，而2017年末至2018年末这段时间中，大盘下跌较为严重，可能造成了相对收益为负的情况
- 2、增持比例较小的分组中，可以看到相比增持较大的分组来讲，其规律性并不强
- 3、增持比例较小的分组中，前期跌幅（相对基准）相较增持大的分组更小
	- 这也符合一般预期，前期跌幅较大，高管认为股价低估买入可能性更大，增持比例可能更大
- 4、在增持比例较大的分组中，基本能获得正向超额收益，2017年除外
- 5、在增持比例较大的分组中，同年所获的相对收益往往要高于增持比例较小的分组
