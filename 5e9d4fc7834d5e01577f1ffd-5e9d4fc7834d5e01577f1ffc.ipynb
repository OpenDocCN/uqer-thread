# 季度三大报表财务因子，选取缺失比例不超过50%的数据

```python
FdmtBSGet = DataAPI.FdmtBSGet(ticker=ticker,secID=u"",reportType=u"A",endDate=u"20191231",
                              beginDate=u"20191231",publishDateEnd=u"",publishDateBegin=u"",endDateRep="",beginDateRep="",
                              beginYear="",endYear="",fiscalPeriod="",field=u"",pandas="1")
FdmtCFGet = DataAPI.FdmtCFGet(ticker=ticker,secID=u"",reportType=u"A",endDate=u"20191231",
                              beginDate=u"20191231",publishDateEnd=u"",publishDateBegin=u"",endDateRep="",beginDateRep="",
                              beginYear="",endYear="",fiscalPeriod="",field=u"",pandas="1")
FdmtISGet = DataAPI.FdmtISGet(ticker=ticker,secID=u"",reportType=u"A",endDate=u"20191231",
                              beginDate=u"20191231",publishDateEnd=u"",publishDateBegin=u"",endDateRep="",beginDateRep="",
                              beginYear="",endYear="",fiscalPeriod="",field=u"",pandas="1")
```

```python
# 选取合并报表
# 选取相同endDate中最大的publishDate
# 选取缺失比例不超过50%的数据
def fun_max_pub(x,df):
    pub_date = max(df.publishDate[(df.endDate == x.endDate)&(df.ticker==x.ticker)])
    return x.publishDate>=pub_date
def select_Report(df, keep_ratio=0.5):
    df = df[df['mergedFlag'] == '1']
    df['last_pub'] = df.apply(lambda x: fun_max_pub(x,df),axis = 1)
    df = df[df['last_pub'] == True]
    df = df[df.columns[df.count()/len(df)>=keep_ratio]]
    df.set_index('ticker',inplace=True)
    df.drop(['secID','publishDate','endDate','endDateRep','partyID','secShortName',
             'exchangeCD','actPubtime','mergedFlag','reportType','fiscalPeriod',
             'accoutingStandards','currencyCD','updateTime','last_pub'],inplace=True,axis=1)
    return df
```

```python
FdmtBSGet = select_Report(FdmtBSGet,0.5)
FdmtCFGet = select_Report(FdmtCFGet,0.5)
FdmtISGet = select_Report(FdmtISGet,0.5)
```

```python
df = pd.merge(df,FdmtBSGet,left_index=True,right_index=True,how='outer')
df = pd.merge(df,FdmtCFGet,left_index=True,right_index=True,how='outer')
df = pd.merge(df,FdmtISGet,left_index=True,right_index=True,how='outer')
```

# 采用均值填充NaN，使用PCA先降维，再使用k-means聚类

```python
df.drop(['tickerBond','tradeDate'],axis=1,inplace=True)
```

```python
df = df.fillna(value=df.mean())
df.drop_duplicates(inplace = True)
```

```python
X_norm = StandardScaler(df.values).copy
```

```python
pipe=Pipeline([('PCA',PCA()),('Kmeans',KMeans())])
```

```python
param_grid = ParameterGrid({'PCA__n_components': [8, 16], 'Kmeans__n_clusters':[8,16]})
labels_dict = {}
scores_dict = {}
for params in param_grid:
    pipe_new = pipe.set_params(**params)
    pipe_new.fit(X_norm)
    labels = pipe_new.predict(X_norm)
    scores = silhouette_score(X_norm, labels , metric='euclidean')
    labels_dict.update({str(params):labels})
    scores_dict.update({str(params):scores})
```

```python
labels_df = pd.DataFrame(labels_dict)
```
