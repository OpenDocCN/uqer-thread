### Tips
---
- 大伙试用风险模型最近常用的就是获取`某一日给定股票的方差协方差矩阵`，可以通过$XFX'+\Delta$计算得到
- 上式中，$X$为因子暴露矩阵，$F$因子协方差矩阵，$\Delta$特质风险矩阵
- 公式看似简单，但是对于数据的存储习惯可不见得大伙都熟悉，这里将一一进行说明

### $X$
---
- 对应的是API是：RMExposureDayGet
- API取出来的就是`最原始的因子暴露矩阵（exposure）`，标准的正态分布

```python
DataAPI.RMExposureDayGet(secID=u"",ticker=u"",tradeDate="20170517",beginDate=u"",endDate=u"",field=u"",pandas="1") .head()
```

### $F$
---
- 对应的API是：RMCovarianceDayGet、RMCovarianceShortGet、RMCovarianceLongGet
- 取出来的数据是`放大10000倍后的年化值`
- 也就是说我们根据真实的factor return估算得到年化后的因子协方差，由于其数值较小，我们就对其进行了放大10000倍的处理，然后存入数据库
- 所以使用时，先要将其缩小10000倍，这样得到的才是预期中的$F$

```python
DataAPI.RMCovarianceDayGet(tradeDate="20170517",beginDate=u"",endDate=u"",Factor=u"",field=u"",pandas="1").head()
```

### $\Delta$
---
- 对应的API是：RMSriskDayGet、RMSriskShortGet、RMSriskLongGet
- 取出的数据是`放大100倍年化后的标准差`
- 使用时，需要先将其缩小100倍，然后取平方变换为对角矩阵，得到的就是预期中的$\Delta$

```python
DataAPI.RMSriskShortGet(secID=u"",ticker=u"",tradeDate="20170517",beginDate=u"",endDate=u"",field=u"",pandas="1").head()
```

### 福利放送
---
- 下面就放送福利，输入日期，即可以获得该日所有股票的方差协方差矩阵
- 计算得到的结果为pd.DataFrame，index和columns均为ticker，value为年化标准差/年化协方差

```python
import numpy as np
import pandas as pd

def get_oneday_cov(date, method='short'):
    '''
    获取某一天所有可取股票的方差协方差矩阵，值为年化后的真实值
    :param date: yyyymmdd
    :param method: 风险模型类型, day/short/long
    :return: pd.DataFrame, index和columns均为ticker
    '''

    expo = DataAPI.RMExposureDayGet(tradeDate=date)
    if len(expo) == 0:
        raise Exception('date should be trading date!!!')
    expo = expo.drop(['tradeDate', 'secID', 'exchangeCD', 'secShortName', 'updateTime'], axis=1).set_index('ticker')

    if method == 'day':
        fcov = DataAPI.RMCovarianceDayGet(tradeDate=date)
        fcov = fcov.drop(['tradeDate', 'FactorID', 'updateTime'], axis=1).set_index('Factor')
        srisk = DataAPI.RMSriskDayGet(tradeDate=date)
        srisk = srisk.drop(['tradeDate', 'secID', 'exchangeCD', 'secShortName', 'updateTime'],
                           axis=1).set_index('ticker')['SRISK']
    elif method == 'short':
        fcov = DataAPI.RMCovarianceShortGet(tradeDate=date)
        fcov = fcov.drop(['tradeDate', 'FactorID', 'updateTime'], axis=1).set_index('Factor')
        srisk = DataAPI.RMSriskShortGet(tradeDate=date)
        srisk = srisk.drop(['tradeDate', 'secID', 'exchangeCD', 'secShortName', 'updateTime'],
                           axis=1).set_index('ticker')['SRISK']
    else:
        fcov = DataAPI.RMCovarianceLongGet(tradeDate=date)
        fcov = fcov.drop(['tradeDate', 'FactorID', 'updateTime'], axis=1).set_index('Factor')
        srisk = DataAPI.RMSriskLongGet(tradeDate=date)
        srisk = srisk.drop(['tradeDate', 'secID', 'exchangeCD', 'secShortName', 'updateTime'],
                           axis=1).set_index('ticker')['SRISK']

    common_ticker = list(set(expo.index) & set(srisk.index))
    expo = expo.loc[common_ticker]
    srisk = srisk[common_ticker]

    fcov = fcov / 10000
    srisk = (srisk / 100) ** 2

    omega = np.matrix(expo.dot(fcov).dot(expo.T).values) + np.matrix(np.diag(srisk.values))
    res = pd.DataFrame(index=expo.index, columns=expo.index, data=omega)
    return res
```
