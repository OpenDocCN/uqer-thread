# 历史百分位计算

```python
#下面getPercentscore(证券代码，指标)函数计算某证券某指标的历史百分位，并画出图形
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
def getPercentscore(tickerstr,fieldstr):
    dfpbpe=DataAPI.MktIdxFactorDateRangeGet(secID=u"",ticker=tickerstr,beginDate='',field="tradeDate,"+fieldstr,pandas="1")
    dfpbpe.index=dfpbpe['tradeDate']
    a=np.zeros(dfpbpe.shape[0])
    for i in range(1,dfpbpe.shape[0]):
        a[i]=stats.percentileofscore(dfpbpe[fieldstr][:i],dfpbpe[fieldstr][i])
    s=pd.Series(a,name=fieldstr+'_p')
    s.index=dfpbpe.index
    s.plot(figsize =(20,5))
    #dfpbpe['pb1_p']=s
    plt.legend(loc='best')
    plt.grid(True)
    print '指数 %s 当前 %s 百分位: %.2f' % (tickerstr,fieldstr[:2],a[i])
getPercentscore("000902","PB2")
```
