机构调研信息作为诸多研报认为的alpha事件之一，当前，机构调研信息一般会在一两个交易日内公布。本文研究该事件是否真的具有超额收益
本文用到的机构调研事件信息从Wind数据库中获得，相关信息以及数据清洗已经完成，保存到jigoudiaoyan.csv这个文件中
下面进行第一步：查看该事件是否具有超额收益，按事件驱动策略的通用策略进行时间回测

```python
import seaborn
from matplotlib import pylab
import matplotlib.pyplot as plt
from matplotlib.pyplot import *
from CAL.PyCAL import *    # CAL.PyCAL中包含font
from datetime import timedelta,datetime
import pandas as pd
from string import zfill
import quartz as qz
raw_df=pd.read_csv('jigoudiaoyan.csv')
##
raw_df['ticker'] = raw_df['ticker'].apply(str).apply(lambda x: zfill(x, 6))
raw_df['secID'] =raw_df['ticker'].map(lambda x: x + '.XSHG' if x[0] == '6' else x + '.XSHE')
##

raw_df_new=raw_df[raw_df['volumn']>=10]  ##volumn表示当日该股票机构调研中的机构家数。这里优选机构家数大于等于10,因为有研报认为调研家数越多，越容易获得更高的超额收益
```

```python
def backtest_time(i, j):  ##i表示机构调研日后几日进场，j表示最大持股时间
    raw_df['t_plus_1'] = raw_df['date'].apply(lambda x: Calendar('China.SSE').advanceDate(x,Period(str(i)+str('B'))).strftime("%Y-%m-%d"))
    start = '2014-08-27'                       # 回测起始时间,大约事件首个交易日
    end = '2017-03-22'                         # 回测结束时间
    benchmark = 'ZZ500'                        # 策略参考标准,中证500
    universe = set_universe('A')  # 证券池，支持股票和基金
    capital_base = 10000000                     # 起始资金
    freq = 'd'                                 # 策略类型，'d'表示日间策略使用日线回测，'m'表示日内策略使用分钟线回测
    refresh_rate = 1                           # 调仓频率，表示执行handle_data的时间间隔，若freq = 'd'时间间隔的单位为交易日，若freq = 'm'时间间隔为分钟
    commission = Commission(buycost=0.0000, sellcost=0.0000)

    def initialize(account):                   # 初始化虚拟账户状态
        account.day_counter = {}  #用于记录持仓天数

    def handle_data(account):                  # 每个交易日的买入卖出指令
    
        for stk in account.avail_security_position:  # 记录持仓的天数
            if account.day_counter.has_key(stk):
                     account.day_counter[stk] += 1
            else:
                     account.day_counter[stk] = 1
            
   ##卖出，最大到j日 卖出         
        for stk in account.security_position:
            if(account.day_counter[stk]>j):   #持仓超过j天，卖出
                     order_to(stk,0)
        
        for stk in (set(account.day_counter.keys())-set(account.security_position.keys())):
                     del account.day_counter[stk]
        
    ##判断当前持股数量        
        current_stock_number=len(account.avail_security_position.keys())
    
    ##买进
        univ_df = raw_df[raw_df['t_plus_1']==account.current_date.strftime('%Y-%m-%d')]
     
        
    
    ##如果当日股票持仓数量少于max_number个，则增仓买入新股票
        
        if len(univ_df )>0: # 当天有事件发生
                univ = univ_df ['secID'].tolist()      
                univ = list(set(univ) - set(account.security_position))  # 已经持有的股票不再重新下单
                need_buy_stock_number=len(univ)##需要新进买入股票的数量
                number=current_stock_number+need_buy_stock_number
            
            
                for stk in univ:
                    order_pct_to(stk,1.0/number)
                for stk in account.security_position:
                    order_pct_to(stk,1.0/number)

       
    bt, perf = qz.backtest(universe=universe, start=start, end=end,
                                initialize=initialize, handle_data=handle_data,
                                capital_base=capital_base, refresh_rate=refresh_rate,
                                freq=freq)
    return perf
```
