# 聊一聊Brinson业绩归因

# 1. 前言
* 好久没发帖子了，最近积分不够用了，发个帖子希望能充个内存。
* 业绩归因不仅仅是多因子模型的最后一个环节，在其他的一些主动模型里也是很重要的分析工具，今天我们就来讲讲Brinson归因

# 2. 单期Brinson归因
首先，我们定义两个组合的收益率$r_{p}$与$r_{b}$ ：
* $r_{p} = \prod_{t=i}^{T}(1 + r_{p_t}) - 1$
* $r_{b} = \prod_{t=i}^{T}(1 + r_{b_t}) - 1$

而：
* $r_{pt} = \sum_{i=1}^{M}w_{p,it}r_{p,it}$
* $r_{bt} = \sum_{i=1}^{M}w_{b,it}r_{b,it}$

我这里定义了M个行业，其中$w_{p, it}$分别表示组合第i个行业的权重与行业的收益率，而$w_{b, it}$则表示基准的第i个行业的权重与行业的收益率。

那么：
$$r_{pt}-r_{bt}=\sum_{i=1}^{M}w_{p,it}r_{p,it} - \sum_{i=1}^{M}w_{b,it}r_{b,it}
=\sum_{i=1}^{M}(w_{p,it}-w_{b,it})(r_{i,bt}-r_{b}) + \sum_{i=1}^{M}w_{p,it}(r_{p,it}-r_{b,it})$$

注意，这样我们就得到了一个简单的单期Brinson归因模型，注意，一般来说，Brinson归因的分解是有三项的，我这里把交叉效应放到了选股效应里面了。

**配置效应**：$\sum_{i=1}^{M}(w_{p,it}-w_{b,it})(r_{i,bt}-r_{b})$。这很好理解，假设基准里面行业i的收益率要高于基准的收益率，而我们组合里面对这个行业又是超配，那么这部分的收益为正；如果要低于基准，我们组合对这个行业又是低配，那么我们这部分的收益也为正。这就是配置的效应
**选择效应**：$\sum_{i=1}^{M}w_{p,it}(r_{p,it}-r_{b,it})$。这也很好理解，假设我们组合里面行业i的收益率要高于基准的收益率，同时，我们也配置了正的权重，那么这部分的收益也为正；同样的道理，如果我们组合里面行业i的收益率要低于基准的收益率，同时，我们也配置了负的权重，那么这部分的收益也为正。只要$w_{p,it}$不为0，这项都不会恒为0，除非我们配置了和基准相同权重的股票或者其他资产。而$w_{p,it}$不为0则表明了我们选择了该行业，这就是选择效应。

# 3. 多期Brinson归因
上面我们介绍了单期的Birnson归因，但是，现实上，我们都需要对一段时间的超额收益进行分析，这里面就牵涉了一个多期归因的问题，因为$r_{p} - r_{b}$并不等于简单的多个单期配置效应与选择效应的相加。

业界一般会引入连接系数来解决这个问题，知乎上也有相应的帖子：
* https://zhuanlan.zhihu.com/p/42893026

# 4. Brinson归因的进一步理解
一般来说，看国内各种分析软件，Brinson归因撑死到这就结束了，我估计魔方大概也是这样。但是对于一个好的金融科技产品来说，如果做到这样我给60分。去年下半年，我去华泰的一个金融科技展，他们做到的大概也是这样，中途还出现了bug，尴尬！

* 假设，我的组合里配了一部分港股，基准是恒生指数，那么这个Brinson业绩归因该怎么做？
* 如果按照A股与港股进行划分，假设A股涨了，港股也涨了，然后就会得到我们组合里A股板块的配置效应是负的，我们不是明明配置了A股了么，那肯定也会贡献正收益了啊，这又该如何理解？
* 其实，我真正想表达的是Brinson归因并不一定就是要基于行业，它的板块可以是区域，现金与股票，行业，甚至可以是直接是股票。还有一个很重要的划分，自己去想。
