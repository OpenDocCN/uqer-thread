# 1. 概况

简单粗暴地，将所有的偏股票型基金分成以下几大类：

![图片注释](http://storage-uqer.datayes.com/56951770228e5b18d9ba2982/627a46f6-d956-11e6-8a64-0242ac140002)

下面拿取数据。需要注意的是，在基金中找量化基金过程中，只是去从名称里面搜索包含特定字段的数据，可能带有一定的主观性，也会导致有些误差。

```python
# 所有基金数据，后面去除分级基金
fund_sec = DataAPI.SecIDGet(assetClass=u"F",pandas="1")
fund_info = DataAPI.FundGet(secID=fund_sec.secID,pandas="1")

fund_leverage = DataAPI.FundLeverageInfoGet(secID=u"",ticker=u"",exchangeCDLeverage=u"XSHG,XSHE",field=u"",pandas="1")
fund_info = fund_info[[x not in fund_leverage.secIDLeverage.values for x in fund_info.secID.values]]

fund_info = fund_info[[x in ['H','E'] for x in fund_info.category]]
fund_info = fund_info[[x not in ['C'] for x in fund_info.operationMode]]
fund_info = fund_info[(fund_info.isFof==0) & (fund_info.isQdii==0) & (fund_info.isGuarFund==0)]
fund_info = fund_info[['标普' not in x and '纳斯达克' not in x and '美国' not in x for x in fund_info.secShortName]]
fund_info = fund_info[['标普' not in str(x) and '纳斯达克' not in str(x) and '美国' not in str(x) for x in fund_info.idxShortName]]
fund_info = fund_info[['标普' not in str(x) and '纳斯达克' not in str(x) and '美国' not in str(x) for x in fund_info.perfBenchmark]]
fund_info = fund_info[['MSCI' not in str(x) and '标准普尔' not in str(x) and '全球' not in str(x) for x in fund_info.perfBenchmark]]
fund_info = fund_info[['欧洲' not in str(x) and '摩根斯坦利' not in str(x) and '富时' not in str(x) for x in fund_info.perfBenchmark]]
fund_info = fund_info[['德国' not in str(x) and '恒生' not in str(x) and '香港' not in str(x) for x in fund_info.perfBenchmark]]
fund_info = fund_info[['海外' not in str(x) and '恒生' not in str(x) and '香港' not in str(x) for x in fund_info.perfBenchmark]]
```

```python
# ----------------------------- 量化基金 -----------------------
# 指数增强
fund_quant_EI_info = fund_info[[('增强' in x or '指数增强' in x) for x in fund_info.secShortName]]

# 绝对收益
fund_quant_hedge_info = fund_info[[('对冲' in x or '绝对收益' in x) for x in fund_info.secShortName]]  

# 量化全部
fund_quant_info = fund_info[[('量化' in x or '多因子' in x or 
                              '对冲' in x or '绝对收益' in x or 
                              '阿尔法' in x or '指数增强' in x) for x in fund_info.secShortName]]
fund_quant_info = fund_quant_info.append(fund_info[fund_info.secID=='229002.OFCN'])
fund_quant_info = fund_quant_info.append(fund_info[fund_info.secID=='202019.OFCN'])
fund_quant_info = fund_quant_info.append(fund_info[fund_info.secID=='001637.OFCN'])
fund_quant_info = fund_quant_info.append(fund_quant_EI_info).append(fund_quant_hedge_info).drop_duplicates('secID')

# 量化主动
fund_quant_active_info = fund_quant_info[[x not in fund_quant_hedge_info.secID.values and 
                                          x not in fund_quant_EI_info.secID.values for x in fund_quant_info.secID]]


# ----------------------------- 非量化基金 -----------------------
# 指数基金
fund_index_info = fund_info[fund_info.indexFund=='I']
fund_index_info = fund_index_info[[x not in fund_quant_info.secID.values for x in fund_index_info.secID]]

# 混合型基金
fund_commingle_info = fund_info[fund_info.category=='H']
fund_commingle_info = fund_commingle_info[[x not in fund_quant_info.secID.values and
                                           x not in fund_index_info.secID.values for x in fund_commingle_info.secID]]

# 普通股票型基金
fund_equ_info = fund_info[fund_info.category=='E']
fund_equ_info = fund_equ_info[[x not in fund_quant_info.secID.values and
                               x not in fund_index_info.secID.values and
                               x not in fund_commingle_info.secID.values for x in fund_equ_info.secID]]

# ----------------------------- 全部基金 -----------------------
fund_all_info = fund_equ_info.append(fund_commingle_info).append(fund_index_info).append(fund_quant_info).drop_duplicates('secID')

fund_secs = {'all': fund_all_info.secID.values, 'equ': fund_equ_info.secID.values, 
             'commingle': fund_commingle_info.secID.values, 'index': fund_index_info.secID.values, 
             'quant': fund_quant_info.secID.values, 'quant_active': fund_quant_active_info.secID.values, 
             'quant_EI': fund_quant_EI_info.secID.values, 'quant_hedge': fund_quant_hedge_info.secID.values}

# ----------------------------- 基金分类整理 -----------------------
category_lv1_dict = {'equ':u'普通股票型', 'index':u'指数基金', 'commingle':u'股票混合', 
                     'quant_active':u'量化基金', 'quant_EI':u'量化基金', 'quant_hedge':u'量化基金'}
category_lv2_dict = {'equ':u'普通股票型', 'index':u'指数基金', 'commingle':u'股票混合', 
                     'quant_active':u'量化主动', 'quant_EI':u'量化指数增强', 'quant_hedge':u'量化对冲'}

fund_all_info = fund_all_info.set_index('secID')

fund_all_info['categoryLevel1'] = np.NaN
fund_all_info['categoryLevel2'] = np.NaN

for x in fund_secs.keys():
    if x == 'all' or x == 'quant':
        continue
    fund_all_info['categoryLevel1'][fund_secs[x]] = category_lv1_dict[x]
    fund_all_info['categoryLevel2'][fund_secs[x]] = category_lv2_dict[x]
    
fund_all_info = fund_all_info.reset_index()
```

## 基金规模变化

统计了过去一年不同分类的基金的规模变化

```python
# 基金历史规模
fund_all_asset = DataAPI.FundAssetsGet(secID=fund_secs['all'], reportDate='', field='')

# 仅统计从09年开始的，基金净值公布日（每个季度最后一天）的数据
date_list = fund_all_asset.groupby('reportDate')['totalAsset'].apply(len).sort_values()
date_list = date_list[date_list.index>='2008-12-31'][date_list>200].sort_index().index
fund_all_asset = fund_all_asset[[x in date_list for x in fund_all_asset.reportDate]]
fund_all_asset.head()

# 只取在fund_all_secs中的基金的规模数据
fund_all_asset = fund_all_asset[[x in fund_secs['all'] for x in fund_all_asset.secID]]
```

# 4. 基金的风险暴露

接下来，我们以上面最后研究的几个规模比较大的、运行时间比较长的量化基金为例，基于公布的持仓数据挖掘这些基金持仓的风险暴露情况；

注意：
- 后面的风险暴露分析基于持仓，而持仓一般只有半年报披露比较完整，所以一下分析不可避免的比较片面，看不到全部，只能做一般参考；
- 另外，需要注意，基金持仓组合的风险，除了本节中讨论的系统性风险，还有股票特异风险；

```python
from lib.Risk import get_risk, get_risk_factors_info, STYLE_FACTORS, INDU_FACTORS
```

计算时，我们采用最常用的多因子模型，将股票收益分解为：
$$r_n = f_c + \Sigma_i X_{ni}f_i + \Sigma_s X_{ns}f_s + u_n$$
其中$f_c$为国家因子收益，$f_i$为行业因子收益，$f_s$为风格因子收益，$X_{ni}$和$X_{ns}$分别为股票对于行业因子i和风格因子s的因子暴露，$u_n$为股票的特异收益率。因此上，如果一个持仓为$w=(w_1,w_2 ... w_N)^T$的组合的收益可以表示为：
$$R_P = \Sigma_n w_n r_n = \Sigma_k f_k (\Sigma_n w_n X_{kn}) + \Sigma_n w_n u_n = \Sigma_k \phi_k f_k + \Sigma_n w_n u_n$$
其中$X_{kn}$为行业因子、风格因子和国家因子合起来的全部风险因子暴露，而$\phi_k=\Sigma_n w_n X_{kn}$为组合对风险因子k的风险暴露，此节中我们重点考察这一风险暴露。对于风格因子和行业因子的说明如下（后面的展示图中直接用因子名简称）：

```python
def sec_exposure(sec, date):
    data = fund_all_hold[(fund_all_hold['secID']==sec) & (fund_all_hold['reportDate']==date)]
    hold_se = data.set_index('holdingSecID')['ratioInNa']
    hold_se = hold_se / 100.0
    exposure = get_risk('exposure', date, 'short')
    exposure = exposure.apply(lambda x: x * hold_se).dropna().sum()
    return exposure

def plot_exposures(sec_exp_df, style=True, industry=False):
    if style:
        ax = sec_exp_df[STYLE_FACTORS].T.plot.bar(figsize=(14,4))
        ax.legend(ncol=min(len(sec_exp_df), 6), loc=9, fontsize=8, prop=font);
        ax.set_xticklabels(STYLE_FACTORS, rotation=0, fontproperties=font);
        ax.set_title(u'风格因子暴露',fontproperties=font);
    
    if industry:
        ax1 = sec_exp_df[INDU_FACTORS[:15]].T.plot.bar(figsize=(14,4))
        ax1.legend(ncol=min(len(sec_exp_df), 6), loc=9, fontsize=8, prop=font);
        ax1.set_xticklabels(INDU_FACTORS[:15], rotation=45, fontproperties=font);
        ax1.set_title(u'行业因子暴露(上)',fontproperties=font);

        ax2 = sec_exp_df[INDU_FACTORS[15:]].T.plot.bar(figsize=(14,4))
        ax2.legend(ncol=min(len(sec_exp_df), 6), loc=9, fontsize=8, prop=font);
        ax2.set_xticklabels(INDU_FACTORS[15:], rotation=45, fontproperties=font);
        ax2.set_title(u'行业因子暴露(下)',fontproperties=font);

        y_max = max(ax1.get_ylim()[1], ax2.get_ylim()[1])
        ax1.set_ylim(0, y_max);
        ax2.set_ylim(0, y_max);
    return 
```
