```python
#3.1pandas基本介绍
import pandas as pd
import numpy as np
s=pd.Series([1,3,6,np.nan,44,1])#创建一个序列
print s#输出序列会发现，pandas当中的元素会带序列号，最终还会输出一个保存格式，通常是64或者32位

dates=pd.date_range('20160101',periods=6)#创建一个日期序列,从2016年1月1日开始，长度为6
print dates

print"矩阵中的一些应用"
df=pd.DataFrame(np.random.randn(6,4),index=dates,columns=['a','b','c','d'])
#随机创建一个6行4列的矩阵，行索引是日期，列索引是abcd
print "df：",df
print "df.dtypes：",df.dtypes#输出每一列的数据类型
print "df.index：",df.index#输出行标签
print "df.columns：",df.columns#输出列标签
print "df.values：",df.values#矩阵每一个元素的value
print "df.describe()：",df.describe()#描述方差、平均值等参数
print "df.T：",df.T#矩阵转置
print "df.sort：",df.sort_index(axis=1,ascending=False)#排序，列标签倒叙排列
print "df.sort_values：",df.sort_values(by='a')#对a列的内容进行升序排列

df1=pd.DataFrame(np.arange(12).reshape((3,4)))#创建一个矩阵，如果不设定他的行列索引，默认0123...
print df1

#创建一个矩阵，以字典的形式，行索引默认是01234，列索引是ABCDEF，内容如下？？？语法错误，但不知道错哪儿
#df2=pd.DataFrame({'A':1.,
#                  'B':pd.Timestamp('20130102)',  
#                  'C':pd.Series(1,index=list(range(4)),dtype='float32')，
#                  'D':np.array([3]*4,dtype='int32'),
#                  'E':pd.Categorical(["test","train","test","train"]),
#                  'F':'foo'})
#df2
#df2.types输出每一列的数据形式

#3.2pandas选择数据
import numpy as np
import pandas as pd

dates=pd.date_range('20130101',periods=6)
df=pd.DataFrame(np.arange(24).reshape((6,4)),index=dates,columns=['A','B','C','D'])
print df
print(df['A'])#输出A列数据
print(df.A)#输出A列数据，输出结果与上一种方式一致
print(df[0:3],df['20130102':'20130104'])#输出第0-2行数据，输出20130102-20130104的数据

#select by label:loc用标签来寻找数据
print (df.loc['20130102'])
print(df.loc[:,['A','B']])
print(df.loc['20130102',['A','B']])

#select by position:iloc用位置寻找数据
print (df.iloc[3])#输出第三行数据
print (df.iloc[3,1])#输出第三行第一列数据
print (df.iloc[3:5,1:3])#切片
print (df.iloc[[1,3,5],1:3])

#mixed selection:ix用标签和位置一起筛选数据
print(df.ix[:3,['A','C']])

#Boolean indexing布尔值筛选
print(df[df.A>8])#找到A列中大于8的内容，输出该数据所在的行所有数据

#3.3pandas设置值
import numpy as np
import pandas as pd

dates=pd.date_range('20130101',periods=6)
df=pd.DataFrame(np.arange(24).reshape((6,4)),index=dates,columns=['A','B','C','D'])

df.iloc[2,2]=1111
print(df)
df.loc['20130101','B']=2222
print(df)
df[df.A>4]=0#先寻找A列中大于4的行，把这些行所有元素改为0
print(df)
df.A[df.A>4]=0#这种情况是将A列中大于4的数据全部改为0
print(df)
df.A[df.A==0]=8#将A列中等于0的值改成8
print(df)
df.B[df.A>4]=0#先寻找A列中大于4的行，修改他在B中的数为0
print(df)

df['E']=np.nan#添加一列空列，E列
print(df)
df['F']=pd.Series([1,2,3,4,5,6],index=pd.date_range('20130101',periods=6))#添加一个F列，自定义一个序列
print(df)

#3.3pandas处理丢失数据
import numpy as np
import pandas as pd

dates=pd.date_range('20130101',periods=6)
df=pd.DataFrame(np.arange(24).reshape((6,4)),index=dates,columns=['A','B','C','D'])
df.iloc[0,1]=np.nan
df.iloc[1,2]=np.nan

print(df)
print(df.dropna(axis=0,how='any'))#how={'any','all'}
#如果某一行中有一个nan，就丢掉该行；how='all'如果改行所有数据都是nan，才丢掉该行
print(df.dropna(axis=1,how='any'))
print(df.dropna(axis=1,how='all'))

print(df.fillna(value=0))#将缺失值填充为0
print(df.isnull())#检查表格中是否有缺失值,有缺失值的位置显示true，其余位置显示false
print(np.any(df.isnull())==True)#表格中只要有缺失值就返回True
```
