```python
from scipy import stats
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import copy
import statsmodels.api as sm
from sklearn.covariance import LedoitWolf
import seaborn as sns
```

```python
class Factor_T():
    def __init__(self,t_step,groupnum,sart_time,keyi,back,*args):
        self.k_k=keyi #第几个因子
        self.back=back #数据向前几日
        self.t_step=t_step #调仓周期
        self.groupnum=float(groupnum) #分组数
        self.sart_time=sart_time #开始时间
        self.__Params()
        self.__loaddata()
        self.__functionDatas()
    def __Params(self):
        self.closeprice=pd.read_csv('closeprice.csv', encoding='GBK')#全市场收盘价
        self.iszz500=pd.read_csv('isZZ500.csv', encoding='GBK')#是否是中证500
        self.maxupdown=pd.read_csv('maxupordown.csv', encoding='GBK')#当日是否涨跌停
        self.iszz500=self.iszz500.T
        self.iszz500.index=self.maxupdown.T.index
        self.iszz500=self.iszz500.T
        self.date_range=self.closeprice.index[0:len(self.closeprice):self.t_step] #整个时间段
        self.IClong=int(120.0/self.t_step) 
        self.rate=(self.closeprice-self.closeprice.shift(self.t_step))/self.closeprice.shift(self.t_step) #周期收益率
        self.day_rate=(self.closeprice-self.closeprice.shift(1))/self.closeprice.shift(1) #日收益率
    def __loaddata(self):#读取相应的因子数据
        self.key=['momentum','pe_ttm','roe_ttm','size','yoyroe','ps']
        self.data={}
        for i in range(len(self.key)):
            self.data[self.key[i]]= pd.read_csv(self.key[i]+'.csv', encoding='GBK')
    def __functionDatas(self):
		self.group,self.d_rate,self.IC0,self.perron_R,self.perron_params,self.perron_corrcef= self.factor_t()
		self.nrate= self.group_rate()
		self.m= self.group_mean_rate()
		self. p_data = self.group_net_value()
    def medianpeak(self,x):
        y=np.copy(x)
        ymedian=np.median(y)
        zmid=np.median(np.abs(y-np.median(y)))
        y[y<=ymedian-5.2*zmid]=(ymedian-5.2*zmid)
        y[y>=ymedian+5.2*zmid]=(ymedian+5.2*zmid)
        return y
    def nondimensionalize(self,x):
        y=np.copy(x)
        ymean=np.mean(y)
        ystd=np.std(y)
        y=(y-ymean)/ystd
        return y
    def maxdro(self,x):
        end = np.argmax((np.maximum.accumulate(x) - x))
        start = np.argmax(x[:end])
        maxdro=1-x[end]/x[start]
        return maxdro
    def factor_t(self):
        perron_R={}
        perron_params={}
        perron_corrcef={}
        IC0={}
        group={}
        ind=0
        for j in range(len(self.date_range)-1):
            if self.date_range[j]-self.back>=0:
                z500=self.iszz500[self.date_range[j]-self.back:self.date_range[j]+1-self.back].T[self.date_range[j]-self.back]
                zz500=z500[z500>0].index
                updown0=self.maxupdown[self.date_range[j]-self.back:self.date_range[j]-self.back+1].T[self.date_range[j]-self.back]
                updown=updown0[updown0==0].index
                index=zz500 & updown
                group[j]={}
                vaild_data=self.data[self.key[self.k_k]][index][self.date_range[j]-self.back:self.date_range[j]+1-self.back].T.dropna().apply(self.medianpeak).apply(self.nondimensionalize)
                sdata=vaild_data
                sdata=sdata.dropna().sort_values(by=self.date_range[j]-self.back,ascending=False)
                perron=[]
                for groupn in range(int(self.groupnum)):
                    group[j][groupn]=sdata.iloc[int(groupn/self.groupnum*len(sdata)):int((groupn+1)/self.groupnum*len(sdata))].index
                    pindex= group[j][groupn] & self.rate[index][self.date_range[j+1]:self.date_range[j+1]+1].T.dropna().index
                    perron.append(self.rate[index][self.date_range[j+1]:self.date_range[j+1]+1].T.dropna().T[pindex].T.mean().values[0])
                    if groupn==self.groupnum-1:
                        ic=[]
                        perron_corrcef[j]=np.corrcoef(perron,np.arange(int(self.groupnum)+1)[1:])[0][1]
                        mm=self.data[self.key[self.k_k]][index][self.date_range[j]-self.back:self.date_range[j]+1-self.back].T.dropna().T
                        m_rate=self.rate[index][self.date_range[j+1]:self.date_range[j+1]+1].T.dropna().T
                        a_index=mm.T.dropna().index & m_rate.T.dropna().index
                        ic.append(np.corrcoef(m_rate[a_index],mm[a_index])[0][1])
                        perr_X=np.arange(int(self.groupnum)+1)[1:]
                        perr_Y=perron-np.mean(perron)
                        perr_X=sm.add_constant(perr_X)
                        model=sm.OLS(perr_Y,perr_X)
                        perr_result=model.fit()
                        perron_R[j]=perr_result.rsquared
                        perron_params[j]=perr_result.params[1]
                        IC0[j]=ic
                if ind==0 and self.date_range[j]-self.back>=0:
                    d_rate=copy.deepcopy(self.day_rate[self.date_range[j]+1:self.date_range[j+1]+1][group[j][self.groupnum-1]].T.mean())
                    ind=1
                elif self.date_range[j]-self.back>=0 and ind==1:
                    d_rate=d_rate.append(self.day_rate[self.date_range[j]+1:self.date_range[j+1]+1][group[j][self.groupnum-1]].T.mean())
        return [group,d_rate,IC0,perron_R,perron_params,perron_corrcef]
    def group_rate(self):
        # 每组每期收益率
        nrate={}
        for i in range(int(self.groupnum)):
            nrate0=[]
            for j in range(len(self.date_range)-1):
                if self.date_range[j]-self.back>=0:
                    nrate0.append(self.rate[self.date_range[j+1]:self.date_range[j+1]+1][self.group[j][i]].T.mean().values[0])
                    nrate[i]=pd.DataFrame(nrate0)
        return nrate
    def group_mean_rate(self):
        # 每组平均收益率
        m=[]
        for i in range(int(self.groupnum)): 
            m.append(self.nrate[i].mean().values[0])
        return m
    def group_net_value(self):
        # 每组单位净值图
        for i in range(int(self.groupnum)):
            p_data0=pd.DataFrame(self.nrate[i])
            if i==0:
                p_data= p_data0
            else:
                p_data= pd.concat([p_data,p_data0],axis=1,join='inner')
        p_data=p_data.T
        p_data.index=np.arange(int(self.groupnum))
        p_data=p_data.T
        p_data=(p_data+1).cumprod()
        return p_data
    def result_show(self):
        # 计算结果
        s_rate=float(len(self.d_rate[self.d_rate>0]))/len(self.d_rate)
        y_rate=np.power((self.p_data[self.groupnum-1][-1:].values[0]),250./len(self.d_rate))-1
        volatility=np.sqrt((250./(len(self.d_rate)-1))*((self.d_rate-self.d_rate.mean())**2).sum())
        X=np.arange(int(self.groupnum)+1)[1:]
        Y=pd.DataFrame(self.m)
        X=sm.add_constant(X)
        model=sm.OLS(Y,X)
        result=model.fit()
        result.params[1]
        sharp_rate=y_rate/volatility
        maxdropd=self.maxdro( (self.d_rate.values[1:]+1).cumprod())
        rr_pp_cc=np.corrcoef(self.m,np.arange(int(self.groupnum)+1)[1:])[0][1]
        rr_mu=pd.DataFrame(self.IC0).T
        rr_volality=np.abs(float(len(rr_mu[0][rr_mu[0]>0]))/len(rr_mu[0])-0.5)
        return [s_rate,y_rate,sharp_rate,maxdropd,result.params[1],rr_pp_cc,rr_volality]
    #return [group,IC0,s_rate,volatility,nrate,m,y_rate,result.params[1],maxdropd]
```

```python
sst={}
rrd={}
mrate={}
n_et={}
for j in range(6):
    sst[j]={}
    rrd[j]={}
    mrate[j]={}
    n_et[j]={}
    for i in range(1,31):
        if i%2==0:
            rr2=Factor_T(i,10,0,j,0)
            sst[j][i]=rr2.factor_t()
            rrd[j][i]=rr2.result_show()
            mrate[j][i]=rr2.group_mean_rate()
            n_et[j][i]=rr2.group_net_value()
```
