# 筹码分布单因子测试

传说中，筹码分布是天狼de陈浩大V的独门绝技，在优矿上搜了一下，也找到了一篇帖子，这里把大概算法贴一下：
N日窗口内Dday移动成本 = 起始价格 + （当日换手率 / 截止Dday日总换手） * （Dday日收盘价格 - 起始价格）
例如：如果要计算7日内某只股票移动成本，则需要循环计算6次，才能计算出7日结束时，相对于7日起始时的成本变动价格。
若下跌时，这个变动价格与起始价格之比越大，说明筹码被套程度越深。

曾经用这个小公式做过多头小策略，股灾1.0后反弹较高，但整个2016年至今均无建树。这里用这个公式做了个alpha单因子测试，效果如下：

```python
# 先导入包
from sklearn import linear_model
import matplotlib.pyplot as plt
from matplotlib import rc
from matplotlib import dates
rc('mathtext', default='regular')
import seaborn as sns
sns.set_style('white')
import datetime as DT
import numpy as np
import pandas as pd
import time
import lib.gra as gra
import scipy.stats as st
from CAL.PyCAL import *    # CAL.PyCAL中包含font
```

```python
# 主函数，写的比较简陋，求轻喷
cal_dates= DataAPI.TradeCalGet(exchangeCD=u"XSHG", beginDate='20080101', endDate='20200101')
trade_dates=cal_dates[cal_dates['isOpen']==1].sort('calendarDate')
# 工作日列表
trade_dates = trade_dates['calendarDate'].tolist()
start = '2015-04-08'
start = DT.datetime.strptime(start,'%Y-%m-%d')
predate=start.strftime('%Y-%m-%d')
predate_index=trade_dates.index(predate)
# period = 5  # 过去N天收益率
# window = 20 # 算法窗口
universe= DynamicUniverse('000906.ZICN')
stock = universe.preview(predate)

tmp_factor = DataAPI.MktEqudAdjGet(secID=stock,ticker=u"",beginDate=trade_dates[predate_index-140],endDate=predate,isOpen="1",field=u"tradeDate,secID,closePrice,turnoverValue",pandas="1")

def get_vp_rate(stock=None,period=5,window=5,start=None,tmp_factor=None):
    predate=start.strftime('%Y-%m-%d')
    predate_index=trade_dates.index(predate)
    secID = tmp_factor['secID']
    tmp_factor['pct_price'] = tmp_factor.groupby('secID')['closePrice'].pct_change(period).replace([-np.inf, np.inf], 0).fillna(value=0)
    tmp_factor=tmp_factor.pivot(index='tradeDate',columns='secID')
    backday_lst = np.arange(0,(period+1),1)
    tmp_group = []
    for backday in [backday_lst[0],backday_lst[-1]]:
        predate_ = trade_dates[predate_index-backday]
        predate_index_ = trade_dates.index(predate_)
        predate_i = trade_dates[predate_index_-window]

        tmp_fac_close = tmp_factor['closePrice'][(tmp_factor.index<=predate_) & (tmp_factor.index>=predate_i)].replace([-np.inf, np.inf], 0).fillna(value=0)
        tmp_fac_Value = tmp_factor['turnoverValue'][(tmp_factor.index<=predate_) & (tmp_factor.index>=predate_i)].replace([-np.inf, np.inf], 0).fillna(value=0)
        tmp_fac_ret = tmp_factor['pct_price'][(tmp_factor.index<=predate_) & (tmp_factor.index>=predate_i)].replace([-np.inf, np.inf], 0).fillna(value=0)
        for x in stock:
            try:
                volPrice = tmp_fac_close[x][0]
                for i in range(window+1):
                    totalTurn = tmp_fac_Value[x].rolling(i+1).sum().values[i]
                    volPrice = volPrice + tmp_fac_Value[x][i]*(tmp_fac_close[x][i] - volPrice) / totalTurn
                    volPrice_rate = volPrice / tmp_fac_close[x][i] - 1.0
                mydata = {'tradeDate':predate_,'secID':x,'volPrice_rate':volPrice_rate,'pct_price':tmp_fac_ret[x][i]}    
                tmp_group.append(mydata)
            except:
                pass
    tmp_df = pd.DataFrame(tmp_group,columns=['tradeDate','secID','volPrice_rate','pct_price']).replace([-np.inf, np.inf], 0).fillna(value=0)
    return tmp_df
```

```python
# 5天筹码成本和5天调仓周期，默认
data = get_vp_rate(stock=stock,start=start,tmp_factor=tmp_factor)
data.head()
```

```python
# 标准化
filter_column = ['secID','tradeDate','pct_price']
data = gra.standard(data,filter_column)
data = gra.sigmoid(data,filter_column)
data
```

```python
# 线性回归看效果
period = 5
current_date=start.strftime('%Y-%m-%d')
currentdate_index=trade_dates.index(current_date)
pre_date = trade_dates[currentdate_index-period]
print current_date,pre_date
# 线性回归
df_factor = data[data['tradeDate']==pre_date]
df_ret = data[data['tradeDate']==current_date]
result = pd.merge(df_factor, df_ret, how='inner', on=['secID'])
# print df_factor,df_ret,result
regr = linear_model.LinearRegression()
regr.fit(result['volPrice_rate_x'].reshape(-1, 1), result['pct_price_y']) # 注意此处.reshape(-1, 1)，因为X是一维的！

a, b = regr.coef_, regr.intercept_
print a
factor = data[data['tradeDate']==current_date]['volPrice_rate'].values[0]
print(a * factor + b)
print(regr.predict(factor))

# 画图
plt.scatter(result['volPrice_rate_x'], result['pct_price_y'], color='blue')
plt.plot(result['volPrice_rate_x'], regr.predict(result['volPrice_rate_x'].reshape(-1,1)), color='red', linewidth=4)
plt.show()
```

默认参数，基本没有方向性，再来看看长成本周期，120天筹码成本，10天调仓周期
