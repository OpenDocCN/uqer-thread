# uqerAPI-Indicator

在写策略时，有时会用到DataAPI，为了方便使用以及防止用到未来数据，我写了如下一个类，分享给大家。

```python
import DataAPI
import CAL.PyCAL as CAL

class IndexData:
    def __init__(self, account, ticker='000001', start = '2012-12-01', end = '2015-07-30' ):
        self.start =start
        self.end = end
        self.account = account
        self.cal = CAL.Calendar('China.SSE')
        self.df = self.index_get(ticker)
        
    def index_get(self, ticker):
        f = lambda x: isinstance(x, str) and x.replace('-','') or x.strftime('%Y%m%d')
        beginDate, endDate = f(self.start), f(self.end)
        df = DataAPI.MktIdxdGet(ticker=ticker, beginDate=beginDate, endDate=endDate, pandas="1")
        return df

    def MA(self, days=5):
        df = self.df
        current_date = self.account.current_date
        current_date = isinstance(current_date, str) and current_date or current_date.strftime('%Y-%m-%d')
        start = self.cal.advanceDate(current_date, '%sB' %(-1-days)).strftime('%Y-%m-%d')
        return df[(df.tradeDate>start) & (df.tradeDate<current_date)]['closeIndex'].mean()
    
    def last(self):
        df = self.df
        current_date = self.account.current_date
        current_date = isinstance(current_date, str) and current_date or current_date.strftime('%Y-%m-%d')
        return df[df.tradeDate==current_date]['preCloseIndex'].values[0]
```
