# Pandas中DataFrame的属性和方法汇总

```python
#DataFrame修改数据方法
from pandas import Series,DataFrame
import pandas as pd
DF={"name":["zhangshan","lisi","wangwu"],"age":[12,13,14],"sex":["male","female","male"]}
f1=DataFrame(DF,index=["one","two","three"],columns=["name","age","sex","interest"])  #此处可添加更多的column
print("---------------数据列表为---------------")
print f1
print("---------------通过Columns大规模修改---------------")
f1["name"]="Dalao"
print f1
print("---------------精准修改1---------------")
f1["name"]["one"]="Zhangshan"
print f1
print("---------------替换整个Columns---------------")
f2=Series(["Zhangshan","Lisi","Wangwu"],index=["one","two","three"])
f1["name"]=f2
print f1
```
