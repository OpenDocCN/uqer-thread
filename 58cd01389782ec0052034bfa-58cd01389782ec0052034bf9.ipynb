# 一种数据类型datetime

datetime是python常见的时间类型，使用方式如下：

```python
from datetime import datetime

print datetime.now()
print datetime(2014,1,4) - datetime(2011,5,4)
print datetime(2013,5,7).month
```

.now就是现在的时间，datetime之间可以做运算，显示year、month、second都是很便捷的

```python
datetime.now()
```

仔细看datetime的结构按照 年 月 日 时 分 秒 微妙 来定义的

- 经常看到datetime的字符类型，咋整？两个方法让你来去自如

```python
date = datetime(2017,3,8,14,23)

date = date.strftime('%Y-%m-%d')
print date
date = datetime.strptime(date,'%Y-%m-%d')
print date
```

strftime从datetime变成string，这里是from time就很好记住了，strptime则相反（不过两个方法的结构不同，挺蛋疼），下面用parse来取代strptime会好看点,另外还有pandas的方法

```python
from dateutil.parser import parse
import pandas as pd

print parse('2013-04-01')
print pd.to_datetime('2013/4/1')
```

- 如何方便地造一组时间序列？有请date界的range大师！

```python
ts = pd.date_range('20160101','20170101',freq='3D')
print ts
```

这里的'3D'就是3天间隔的意思，这里引申出一个时间序列**频率及偏移量**的概念。

```python
#我是频率
print pd.date_range('20160101','20170101',freq='4h30min')
print pd.date_range('20160101','20170101',freq='WOM-3Fri')#week of month = WOM
print '='*100
#我是偏移
from pandas.tseries.offsets import *
print ts.shift(2,freq='M')
offset = BDay()#工作日
print offset.rollback(datetime.now())
```

- 如何做时间序列的采样？这里有个resample

这里要做resample的时候需要更改uqer数据类型，这是个坑，当然大部分帖子都已经做了这方面的转换了

```python
data = DataAPI.MktEqudGet(secID=u"",ticker=u"000001",beginDate=u"20170101",endDate=u"20170301",isOpen="",field=u"tradeDate,openPrice,highestPrice,lowestPrice",pandas="1")
data['tradeDate'] = pd.to_datetime(data.tradeDate)
data.set_index('tradeDate',inplace=True)
print data.head()
```

```python
print data.resample('M',how='sum',fill_method=None,closed='right',label='left')
print data.resample('H',how='mean',fill_method='ffill',closed='right',label='right').head()
```

resample中基本的参数包括：1）频率；2）聚合的方法；3）填充空值的方法；4）闭合的方式；5）tradeDate使用的标签；具体细节大家自己试试，很容易就能看出区别，最后给个图示，方法简单明了
