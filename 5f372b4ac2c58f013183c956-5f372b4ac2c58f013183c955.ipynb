# 20200218移动筹码分布

移动筹码分布是如何计算以及计算原理如下：
　　筹码分布就是将历史上在每个价位成交的量叠加起来，并以此来判断当前市场上所有流通股的持仓成本。
　　当然历史上成交中的一部分会在后面的交易日中被抛出，也就是说不能简单地将以前的成交累积到现在，而应该有一定的衰减。这个衰减的比例也就是每天的换手率。
　　比如说，1000万的盘子，前天均价为10元，成交量为200万，也就是20%换手率；昨天以均价11元又成交300万，也就是30%换手率；那前天的200万成交量怎么样了呢？成本分析假定，前天

的200万在昨天也以11元被30%换手了，那么，前天以10元成交的成交量还剩了200*(1-30%)=140万；若今天以均价12元又成交了400万，同理可算，现在的筹码分布是：10元筹码为200*(1-30%)*(1-40%)=84万，11元的筹码为300*(1-40%)=180万，12元的筹码是400万。

```python
def get_chouma(ticker="600668",beginDate=u"20150101",endDate="20200214"):
    data1=DataAPI.EquGet(ticker=ticker,equTypeCD=u"A",listStatusCD=u"",
                         field=u"ticker,secShortName,nonrestFloatShares",pandas="1")
    #print "nonrestFloatShares=",data1.nonrestFloatShares.values
    data=DataAPI.MktEqudAdjGet(ticker=ticker,tradeDate=u"",beginDate=beginDate,                        endDate=endDate,isOpen="1",field=u"ticker,tradeDate,closePrice,turnoverVol,turnoverValue,turnoverRate",pandas="1")
    data['chouma']=0
    #data['nonrestFloatShares']=data.apply(lambda x:x.turnoverVol/x.turnoverRate, axis = 1)
    data0=DataAPI.EquShareGet(ticker=ticker,partyID=u"",
                                        beginDate=u"",endDate=endDate,field=u"",pandas="1")
    #20200515注意 这里要求【0】,即可每列nonrestFloatShares，均取到endDate的最新nonrestFloatShares的值
    data['nonrestFloatShares']=data0.head(1).nonrestFloatShares[0]
    data=data.sort(["tradeDate"],ascending=False)
    data=data.reset_index(drop=True)
#data=data.head(4)
    alpha=1
    data.chouma[0]=data.turnoverVol[0]*alpha
    for i in range(1,len(data)):    
            alpha=1  
            for j in range(0,i):
                #print "假设昨天的交易量以今天的换手率【的0.382】被换手:\nalpha=alpha*(1-0.5*data.turnoverRate[j])"
                alpha=alpha*(1-0.382*data.turnoverRate[j])
                #alpha=alpha*(1-data.turnoverRate[j])
                #print ("i=%d,j=%d,alpha= %f,turnoverRate[j]=%f" %(i,j,alpha,data.turnoverRate[j]))
            #print "\n"
            data.chouma[i]=data.turnoverVol[i]*alpha
    #print data.head()
    #print data.tail()
    sum_chouma=0
    for i in range(0,len(data)):
            sum_chouma=sum_chouma+data.chouma[i]
            #print "i=%d,sum_chouma=%f:data.chouma[i]=%f" %(i,sum_chouma,data.chouma[i])
            if (sum_chouma>data1.nonrestFloatShares.values):
                #print "i=%d,sum_chouma=%f:data.chouma[i]=%f" %(i,sum_chouma,data.chouma[i])
                break
    df=data.head(i+1)
    return df
ticker="600668"
beginDate="20190101"
endDate="2020-03-13"
df=get_chouma(ticker=ticker,beginDate=beginDate,endDate=endDate)
print "假设昨天的交易量以今天的换手率【的0.382】被换手:\nalpha=alpha*(1-0.382*data.turnoverRate[j])"
print "得到全部筹码全换手的区间数据"
print df.head(2)
print df.tail(2)
newbeginDate=df.tradeDate[len(df)-1]
newendDate=df.tradeDate[0]
#print newbeginDate
df['day_meanprice']=df.apply(lambda x:x.turnoverValue*1.0/x.turnoverVol, axis = 1)
df=df.sort(["day_meanprice"],ascending=True)
df=df.reset_index(drop=True)
sum_vol=df.turnoverVol.sum()
#print "sum_vol:",sum_vol
tmp1_vol=0
#print df
for i1 in range(0,len(df)):
            tmp1_vol=tmp1_vol+df.turnoverVol[i1]
            q005=df.day_meanprice[0]
            if (tmp1_vol<=sum_vol*0.05) and (tmp1_vol+df.turnoverVol[i1+1]>sum_vol*0.05):
                residue_vol=sum_vol*0.05-tmp1_vol
                residue_Value=residue_vol*df.day_meanprice[i1+1]
                q005=(residue_Value+df.turnoverValue[i1])/(df.turnoverVol[i1]+residue_vol)
                print "i1=%d,tmp1_vol=%f:df.day_meanprice[i1]=%f,df.turnoverVol[i1]=%f" %(i1,tmp1_vol,df.day_meanprice[i1],df.turnoverVol[i1])
                break
tmp2_vol=0 
for i2 in range(0,len(df)):
            tmp2_vol=tmp2_vol+df.turnoverVol[i2]
            if (tmp2_vol<=sum_vol*0.95) and (tmp2_vol+df.turnoverVol[i2+1]>sum_vol*0.95):
                residue_vol=sum_vol*0.95-tmp2_vol
                residue_Value=residue_vol*df.day_meanprice[i2+1]
                q095=(residue_Value+df.turnoverValue[i2])/(df.turnoverVol[i2]+residue_vol)
                print "i2=%d,tmp1_vol=%f:df.day_meanprice[i2]=%f,df.turnoverVol[i2]=%f" %(i2,tmp2_vol,df.day_meanprice[i2],df.turnoverVol[i2])
                break
print "len(df)=%d,i1=%d,i2=%d" %(len(df),i1,i2)
#q005=df[['day_meanprice']].quantile(0.05)
#q095=df[['day_meanprice']].quantile(0.95)
jizhongdu=(q095-q005)*1.0/(q095+q005)
print "%s %s %s %d 集中度：%.2f" %(ticker,newbeginDate,newendDate,len(df),jizhongdu)
#print "90%%成本%f-%f,jizhongdu=%f" %(q005,q095,jizhongdu)
```

```python
def get_jizhongdu(ticker="600668",beginDate=u"20150101",endDate="20200214"):
    df=get_chouma(ticker=ticker,beginDate=beginDate,endDate=endDate)
    newbeginDate=df.tradeDate[len(df)-1]
    newendDate=df.tradeDate[0]
    df['day_meanprice']=df.apply(lambda x:x.turnoverValue*1.0/x.turnoverVol, axis = 1)
    df=df.sort(["day_meanprice"],ascending=True)
    df=df.reset_index(drop=True)
    sum_vol=df.turnoverVol.sum()
    #print "sum_vol:",sum_vol
    tmp1_vol=0
    #print df
    for i1 in range(0,len(df)):
            tmp1_vol=tmp1_vol+df.turnoverVol[i1]
            q005=df.day_meanprice[0]
            if (tmp1_vol<=sum_vol*0.05) and (tmp1_vol+df.turnoverVol[i1+1]>sum_vol*0.05):
                residue_vol=sum_vol*0.05-tmp1_vol
                residue_Value=residue_vol*df.day_meanprice[i1+1]
                q005=(residue_Value+df.turnoverValue[i1])/(df.turnoverVol[i1]+residue_vol)
                print "i1=%d,tmp1_vol=%f:df.day_meanprice[i1]=%f,df.turnoverVol[i1]=%f" %(i1,tmp1_vol,df.day_meanprice[i1],df.turnoverVol[i1])
                break
    tmp2_vol=0 
    for i2 in range(0,len(df)):
            tmp2_vol=tmp2_vol+df.turnoverVol[i2]
            if (tmp2_vol<=sum_vol*0.95) and (tmp2_vol+df.turnoverVol[i2+1]>sum_vol*0.95):
                residue_vol=sum_vol*0.95-tmp2_vol
                residue_Value=residue_vol*df.day_meanprice[i2+1]
                q095=(residue_Value+df.turnoverValue[i2])/(df.turnoverVol[i2]+residue_vol)
                print "i2=%d,tmp1_vol=%f:df.day_meanprice[i2]=%f,df.turnoverVol[i2]=%f" %(i2,tmp2_vol,df.day_meanprice[i2],df.turnoverVol[i2])
                break
    print "len(df)=%d,i1=%d,i2=%d" %(len(df),i1,i2)
    #q005=df[['day_meanprice']].quantile(0.05)
    #q095=df[['day_meanprice']].quantile(0.95)
    jizhongdu=(q095-q005)*1.0/(q095+q005)
    print "%s %s %s %d 集中度：%.2f" %(ticker,newbeginDate,newendDate,len(df),jizhongdu)
    #print "90%%成本%f-%f,jizhongdu=%f" %(q005,q095,jizhongdu)
    allturnover_days=len(df)
    return allturnover_days,round(jizhongdu,2)
ticker="600668"
beginDate="20190101"
endDate="2020-03-13"
allturnover_days,jizhongdu=get_jizhongdu(ticker=ticker,beginDate=beginDate,endDate=endDate)
```
