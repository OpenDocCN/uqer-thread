```python
#循环中常伴的两种语法：break，continue,一般与if同时使用
#break：在循环中，break可以提前结束该循环。
for i in range(10):
    if i >=5:
        break
    print(i)
```

```python
#continue:在循环中，continue可以跳过当前循环，进行下一次的循环
for i in range(10):
    if i %2==0:
        continue
    print(i)
```

```python
a =[]
for i in range(10):
    if i%2 !=0:
        a.append(i)
a
```

```python
#列表表达式
a= [i for i in range(10) if i%2 != 0]
a
```

5.函数
减少部分重复代码；抽象方式

```python
def my_sum():
    cnt = 0
    for i in range(101):
        cnt +=i
    print(cnt)

my_sum()
```

```python
def my_sum(n):
    cnt = 0
    for i in range(n):
        cnt +=i
    print(cnt)
    return cnt    #函数返回，如果不写，会默认返回空值

cnt = my_sum(101)
cnt
```
