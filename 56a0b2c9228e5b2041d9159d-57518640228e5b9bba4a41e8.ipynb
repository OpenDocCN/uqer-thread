# 【择时研究】成交量热度——高抛低吸不如站在“炒作”的风口上

## 【序言】

- 马尔科姆·格拉德威尔在他的著作《引爆点》中分析了事物流行的要素和过程，他认为，思想、行为、信息及产品常会像传染病暴发一样迅速传播。在股市中，也有一种东西具备引爆点的条件，那就是市场情绪。从粗分类上看，市场情绪无非可以分为三种：乐观、中性和悲观。当一种情绪启动（说人话就是“炒作”），市场由于羊群效应在接下来的几天会出现短期的正反馈和跟风现象，市场情绪在短期后市得以传播和反映到市场。如果能捕捉到市场短期的乐观情绪，那么便可以悄悄的入市，待市场炒作起来之后，全身而退深藏功与名；若捕捉到市场短期的悲观情绪，便可以选择观望或者割肉逃跑，以防更大损失的可能性。

- 本文试图从成交量下手去寻找可以判断市场情绪启动的信号，并研究根据信号及时跟风入市，再持有合适的期限后获取收益的情况。

- 将成交量热度指标分为成交量正向热度和成交量反向热度，其中前者代表成交量短期突增带来的信号，后者代表成交量短期萎靡带来的信号，我们将它们均作为某种市场情绪的启动信号。

## 【成交量热度指标】

- 股票的两个重要构成元素包括价格和成交量，成交量通常蕴含重要的交易信息，它直接反映交易行为，体现当前市场的交易热度。

- 成交量热度是通过标的成交量的大小变化反映标的在当前市场活跃程度的变化。

- 将标的上一个交易日的成交量在过去一段历史时间窗口的排名作为成交量热度指标，若排名靠前则认为标的在当前市场热度高，正处于短期放量的“炒作”时期，是正向热度；若排名靠后则认为该标的当前市场热度低，处于“萎靡”时期，是反向热度。

- 我们对正向热度和反向热度的信号分别作择时检验，并研究其所给能出的信息量。

## 【正向指标检验】

- 选取沪深300指数作为标的，研究大盘在成交量短期放量后市的表现。

- 某日信号生成方法
    - step1 在当前日，计算前一日成交量在一段历史窗口的排名。
    - step2 若排名靠前面30%，则给出买入信号，按照开盘价格买入。


```python
stock = '000300'
stkData = origData[origData.ticker == stock]
# stkData['rank'] = stkData['tradeDate']
histLength = 20
signal = []
for index in stkData['tradeDate']:
    tmpData = stkData[stkData.tradeDate < index][-histLength:]
    if len(tmpData)< histLength:
        signal.append(0)
    else:
        tmpData = tmpData['turnoverVol'].tolist()
        rank = sorted(cp.deepcopy(tmpData),reverse = True).index(tmpData[-1])
        
        if rank < histLength/3:
            signal.append(1)
        else:
            signal.append(0)
        
stkData['signal'] = signal
```

- 持有期限
    - 持有期限从2个交易日到20个交易日，研究其胜率变化

```python
netAsset = (stkData['closeIndex']/stkData['closeIndex'].tolist()[0]-1).tolist()
openDates = [index+1 for index in range(20)]
winRates = {}
for openDate in openDates:
    sigRet = []
    for index in stkData['tradeDate']:
        tmpData = stkData[stkData.tradeDate == index]
        if tmpData['signal'].tolist()[0] == 1:
            buyPrice = tmpData['closeIndex'].tolist()[0]
            sellPrice = stkData[stkData.tradeDate>index][:openDate].tail(1)['closeIndex'].tolist()[0]
            sigRet.append(sellPrice/buyPrice*1.0-1)
        else:
            sigRet.append(0)
    winRates[openDate] = round(1.0*len([index for index in sigRet if index > 0])/(len(sigRet)-sigRet.count(0)),2)
```

- 图示“胜率-持有期限”之间的关系

```python
plt.figure(figsize = (14,8))
plt.title('winRate-holdingPeriod',fontsize = 30)
plt.plot(winRates.keys(),winRates.values(),'r-*',markersize = 10)
plt.legend(['win rate'])
plt.tick_params(labelsize = 20)
```

- 很有趣的是，持有期从1天到一个月，胜率每天都大于50%！且持有期限在三周以内胜率几乎呈现一种随着持有期限的增加而稳步上升的趋势，持有两到三周的操作方法得到的胜率最高。

- 我们对于持有期为15天的情况做收益曲线回测，首先计算每个发出的买信号持有15天后得到的收益率回报

```python
openDate = 15
sigRet = []
for index in stkData['tradeDate']:
    tmpData = stkData[stkData.tradeDate == index]
    if tmpData['signal'].tolist()[0] == 1:
        buyPrice = tmpData['openIndex'].tolist()[0]
        sellPrice = stkData[stkData.tradeDate>index][:openDate].tail(1)['closeIndex'].tolist()[0]
        sigRet.append(sellPrice/buyPrice*1.0-1)
    else:
        sigRet.append(0)
stkData['sigRet'] = sigRet
```

- 按照复利全仓买入的原则，我们按照该原则过滤掉一些收益数值，具体的过滤方法为

    - 从当天往算推算一个持有期，若持有期内存在收益数值，则表明当前没有现金投入到新的信号中，此时应该讲当前的收益数值置0。
    
    - 过滤是为了可操作，但同时带来的一个问题是策略的收益将和时间起点相关。当然，如果该信号带来的收益是有效且平稳的，那么将与时间起点无关，待考证。

```python
realRet = sigRet
for index in range(len(sigRet)):
    if index < openDate:
        if sum(sigRet[:index])!=0:
            realRet[index] = 0
    else:
        if sum(sigRet[index-openDate:index]) != 0:
            realRet[index] = 0
stkData['realRet'] = realRet
```

- 按照复利计算收益曲线并与基准做对比

```python
netAsset = (stkData['closeIndex']/stkData['closeIndex'].tolist()[0]).tolist()
ptfAsset = [1]
for index in realRet[1:]:
    ptfAsset.append(ptfAsset[-1]*(1+index))

plt.figure(figsize = (14,8))
plt.title('returns & benchmark',fontsize = 30)
plt.plot(netAsset,'b',markersize = 10)
plt.plot(ptfAsset,'r',markersize = 10)
plt.legend(['netAsset','ptfAsset'])
plt.tick_params(labelsize = 20)
```

- 做对冲的超额收益

```python
difAsset = [ptfAsset[index] - netAsset[index] for index in range(len(netAsset))]
plt.figure(figsize = (14,8))
plt.title('hedge returns',fontsize = 30)
plt.plot(difAsset,'b',markersize = 10)
plt.tick_params(labelsize = 20)
plt.legend(['hedge returns'])
plt.plot(difAsset)
```

- 5年回测，60%的对冲收益，平均年化12%，以大盘为标的，可投资相关宽基指数基金，相对来说是不错的表现。

## 【反向指标检验】

- 选取沪深300指数作为标的，研究大盘在成交量短期缩量后市的表现。

- 某日信号生成方法
    - step1 在当前日，计算前一日成交量在一段历史窗口的排名。
    - step2 若排名靠后面30%，则给出买入信号，按照开盘价格买入。

```python
####单只股票表现
stock = '000300'
stkData = origData[origData.ticker == stock]
# stkData['rank'] = stkData['tradeDate']
histLength = 20
signal = []
for index in stkData['tradeDate']:
    tmpData = stkData[stkData.tradeDate < index][-histLength:]
    if len(tmpData)< histLength:
        signal.append(0)
    else:
        tmpData = tmpData['turnoverVol'].tolist()
        rank = sorted(cp.deepcopy(tmpData),reverse = True).index(tmpData[-1])
        
        if rank > 2.0*histLength/3:
            signal.append(1)
        else:
            signal.append(0)
        
stkData['signal'] = signal
```

- 持有期限
    - 持有期限从2个交易日到20个交易日，研究其胜率变化

```python
openDates = [index+1 for index in range(20)]
winRates = {}
for openDate in openDates:
    sigRet = []
    for index in stkData['tradeDate']:
        tmpData = stkData[stkData.tradeDate == index]
        if tmpData['signal'].tolist()[0] == 1:
            try:
                buyPrice = tmpData['closeIndex'].tolist()[0]
                sellPrice = stkData[stkData.tradeDate>index][:openDate].tail(1)['closeIndex'].tolist()[0]
                sigRet.append(sellPrice/buyPrice*1.0-1)
            except:
                sigRet.append(0)
                pass
        else:
            sigRet.append(0)
    winRates[openDate] = round(1.0*len([index for index in sigRet if index > 0])/(len(sigRet)-sigRet.count(0)),2)
```

- 画出“胜率-持有期限”图

```python
plt.figure(figsize = (14,8))
plt.title('winRate-holdingPeriod',fontsize = 30)
plt.plot(winRates.keys(),winRates.values(),'r-*',markersize = 10)
plt.tick_params(labelsize = 20)
plt.legend(['win rate'])
```

- 可见，后30%的股票随后几日的表现和前30%正好是负相关的关系，且平均胜率低于50%。
