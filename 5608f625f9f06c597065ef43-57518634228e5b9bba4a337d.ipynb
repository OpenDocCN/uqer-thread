# 50ETF Option Vol Smile Curve

```python
from CAL.PyCAL import *
from pandas import Series, DataFrame, concat
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import time
from scipy.stats import norm
from scipy.optimize import fsolve
```

```python
def OptBSV(Forward,Strike,DsctRate,Vol,Term,phi):
    d1=np.log(Forward/Strike)/Vol/np.sqrt(Term)+0.5*Vol*np.sqrt(Term)
    d2=d1-Vol*np.sqrt(Term)
    return phi*(Forward*norm.cdf(phi*d1,0,1)-Strike*norm.cdf(phi*d2,0,1))*np.exp(-DsctRate*Term)
```

```python
DsctRate=0.03 #zero-coupon rate
SpotDate=Date(2015,9,8) #val date

SpotDateStr=SpotDate.toISO().replace('-','')
OptMkt=DataAPI.MktOptdGet(tradeDate=SpotDateStr,field=[u"optID","tradeDate","closePrice"],pandas="1")
OptInfo=DataAPI.OptGet(optID=map(str,OptMkt.optID),field=[u"optID",u"varTicker",u"contractType",u"strikePrice",u"listDate",u"expDate",],pandas="1")
OptInfo=OptInfo.sort("optID").set_index("optID").join(OptMkt.sort("optID").set_index("optID"))

Underly=DataAPI.MktFunddGet(ticker=u"510050",tradeDate=SpotDateStr,field=u"closePrice",pandas="1")
OptInfo=OptInfo[OptInfo.varTicker==510050]
OptInfo["spotPrice"]=Underly.closePrice[0]
expDate=Series(OptInfo.expDate.values.ravel()).unique().tolist()
OptInfo=OptInfo[OptInfo.expDate==expDate[1]]
callOptInfo=OptInfo[OptInfo.contractType=="CO"]
putOptInfo=OptInfo[OptInfo.contractType=="PO"]
callOptInfo=callOptInfo[callOptInfo.closePrice>callOptInfo.spotPrice-callOptInfo.strikePrice].sort("strikePrice")
putOptInfo=putOptInfo[putOptInfo.closePrice>putOptInfo.strikePrice-putOptInfo.spotPrice].sort("strikePrice")

OptInfo=pd.DataFrame([])
OptInfo["strike"]=np.intersect1d(callOptInfo.strikePrice,putOptInfo.strikePrice)
OptInfo["call"]=np.zeros_like(OptInfo.strike)
OptInfo["put"]=np.zeros_like(OptInfo.strike)
for i in range(0,len(OptInfo.strike)):
    OptInfo["call"][i]=np.array(callOptInfo.closePrice[callOptInfo.strikePrice==OptInfo.strike[i]])[0]
    OptInfo["put"][i]=np.array(putOptInfo.closePrice[putOptInfo.strikePrice==OptInfo.strike[i]])[0]
trade=np.array(callOptInfo["tradeDate"])[0]
expiry=np.array(callOptInfo["expDate"])[0]
term=(Date.parseISO(expiry)-Date.parseISO(trade))/365.
spot=np.array(callOptInfo["spotPrice"])[0]
OptInfo["forward"]=(OptInfo["call"]-OptInfo["put"]+OptInfo["strike"])*np.exp(DsctRate*term)


OptInfo["callVol"]=np.zeros_like(OptInfo.strike)
OptInfo["putVol"]=np.zeros_like(OptInfo.strike)
for i in range(0,len(OptInfo.strike)):
    OptInfo["callVol"][i]=fsolve(lambda x:OptBSV(OptInfo["forward"][i],OptInfo["strike"][i],DsctRate,x,term,1)-OptInfo["call"][i], 0.2)
    OptInfo["putVol"][i]=fsolve(lambda x:OptBSV(OptInfo["forward"][i],OptInfo["strike"][i],DsctRate,x,term,-1)-OptInfo["put"][i], 0.2)
    
OptInfo=OptInfo.set_index("strike")
OptInfo[["callVol","putVol"]].plot(figsize=(12,7))
plt.xlabel("strike")
plt.ylabel("vol")
```
