```python
from CAL.PyCAL import *
import pandas as pd
today=Date.todaysDate()
cal=Calendar("China.SSE")
yesterday=cal.advanceDate(today,"-1B",BizDayConvention.Unadjusted)
universe=set_universe("A")
data=DataAPI.MktTickRTSnapshotGet(securityID=universe,field=["shortNM","lastPrice","bidBook","askBook"])
data.set_index("ticker",inplace=True)
preClp=DataAPI.MktEqudGet(secID=universe,beginDate=yesterday.strftime("%Y%m%d"),endDate=yesterday.strftime("%Y%m%d"),field=["ticker","closePrice"])
preClp.set_index("ticker",inplace=True)
data=pd.merge(data,preClp,left_index=True,right_index=True,how="inner")
data.reset_index(inplace=True)

def change(x):
    secID=str(x[0])+"."+x[1]
    return secID
data["secID"]=data.apply(change,axis=1)

#跌停板封板资金
dnlimit=data[(data["bidBook_volume1"]==0)&(data["askBook_volume1"]>0)]
dnlimit["bangvol"]=dnlimit["askBook_price1"]*dnlimit["askBook_volume1"]/10000

#涨停板封板资金
uplimit=data[(data["bidBook_volume1"]>0)&(data["askBook_volume1"]==0)]
uplimit["bangvol"]=uplimit["bidBook_volume1"]*uplimit["bidBook_price1"]/10000
uplimit=uplimit[["shortNM","secID","closePrice","bidBook_price1","bidBook_volume1","bangvol"]]
# uplimit.columns=["简称","代码","前收","最新价","封单量","封单金额（万）"]
# uplimit.reset_index(inplace=True)
# uplimit.drop("index",axis=1,inplace=True)

#计算涨停板封板量与流通总股本百分比
nflist=DataAPI.EquGet(equTypeCD="A",field=["secID","nonrestfloatA","listDate"])
uplimit.set_index("secID",inplace=True)
nflist.set_index("secID",inplace=True)
uplimit=pd.merge(uplimit,nflist,left_index=True,right_index=True,how="inner")
uplimit["nfrate"]=uplimit["bidBook_volume1"]/uplimit["nonrestfloatA"]*100
uplimit.sort("nfrate",ascending=False,inplace=True)
#剔除最近一个月上市的新股
uplimit=uplimit[uplimit.listDate<"2016-10-07"]
uplimit

```
