* ## 全国百城平均房价

```python
import lib.lppl as lppltool
from matplotlib import pyplot as plt
import datetime
import numpy as np
import pandas as pd
import seaborn as sns
sns.set_style('white')

limits = ([9.5, 10], [-1, -0.1], [80, 120], [.1,.9], [-1,1], [15,18], [0, 2*np.pi])
x = lppltool.Population(limits, 20, 0.3, 1.5, .05, 4)
for i in range (2):
	x.Fitness()
	x.Eliminate()
	x.Mate()
	x.Mutate()
	
x.Fitness()	
values = x.BestSolutions(3)
for x in values:
	print x.PrintIndividual()
```

请看上面的数据，三个fit的critical time 都是120左右，相对来说，就是从现在起50个月，也就是4年之后。。。
也就是暂时总体房价没有达到泡沫的峰值。
拟合的fitness也在0.03左右。

```python
data = pd.DataFrame({'Date':values[0].getDataSeries()[0],'Index':values[0].getDataSeries()[1],'Fit1':values[0].getExpData(),'Fit2':values[1].getExpData(),'Fit3':values[2].getExpData()})
data = data.set_index('Date')
data.plot(figsize=(14,8))
```

* ## 深圳房价

```python
import lib.shenzhen as lppltool
from matplotlib import pyplot as plt
import datetime
import numpy as np
import pandas as pd
import seaborn as sns
sns.set_style('white')

limits = ([11, 12.5], [-1, -0.1], [75, 100], [.1,.9], [-1,1], [5,15], [0, 2*np.pi])
x = lppltool.Population(limits, 20, 0.3, 1.5, .05, 4)
for i in range (2):
	x.Fitness()
	x.Eliminate()
	x.Mate()
	x.Mutate()
	
x.Fitness()	
values = x.BestSolutions(3)
for x in values:
	print x.PrintIndividual()
    
    
```
