# `股息率因子的研究意义`
#### 股息率，即每股派息除以最新每股价格，衡量的是上市公司可以为股东提供稳定现金流的能力强弱。当前，投资者需要更加关注高股息率股票，主要是以下原因：
## `量化因子失效`
#### 从2016年年末开始，过去10年收益较好的市值因子，流动性因子以及反转因子提供的超额收益逐渐缩小甚至出现了较大的回撤，而恰恰从2016年年末开始，价值因子才开始领先市场并提供了长期稳定的超额收益。这与大环境以及政策导向有关，而股息率因子是一个信息含量丰富的价值因子，因此需要更加细致的研究。
## `股息率的内涵`
#### 1. “能赚钱”，有股息的前提是公司有所应力，亏钱的公司无法分红；
#### 2. “经营预期好”，不同于转股送股，分红是公司拿出真金白银分给投资者，因为如果公司经营状况较差或者行业增长停滞，企业不会拿出大笔资金进行分红；
#### 3. “足够便宜”，哪怕企业有大量的现金分红，但是股价高企的话，其股息率依旧会偏低。
## `本文解决的三个问题`
#### 1. 什么样的公司会分红或者分红更多？
#### 2. 什么样的公司适合用股息率定价？
#### 3. 什么样的行业适合用股息率定价？

# `什么公司适合用股息率定价`
#### 一般认为，对于不同生命周期的上市公司，其估值方式是有区别的
#### 1. 处于初创阶段的公司看重其业务前景和成长性。
#### 2. 处于成长阶段的公司看重其成长性和业绩兑现。
#### 3. 处于成熟阶段的公司看重其盈利能力和稳定性。
#### 从股息率的定义来看，它更看重公司的稳定盈利能力，因此猜想成熟阶段的公司更加适合用股息率进行估值。

## `行业入选股则`
#### 由于行业平均市值，平均ROE和行业平均未分配利润占比在某种程度上有较强的相关性，因此从四个指标中仅选择行业平均未分配利润占比以及行业内股息率Spread作为筛选指标，则行业选入规则如下：
#### 1. 每一个调仓日，计算每一个申万一级行业的行业平均未分配利润占比和行业内股息率Spread;
#### 2. 分别计算每个行业在两个指标上面的排名，根据两个排名得到调仓日申万一级行业的综合排名：
#### 3. 选入综合排名前10 的申万一级行业。
![行业选入规则](http://img.027cgb.cn/20170505/20175195002016273179.png)

```python
def stock_list(current):
    '''
    得到关于时间的分红列表股票及相关属性聚合
    '''
    universe1=DynamicUniverse('A')
    current=current
    y_day=cal.advanceDate(current,Period('-1B')).strftime("%Y-%m-%d")
    start=cal.advanceDate(current,Period('-2Y')).strftime("%Y-%m-%d")
    #start_RPP=cal.advanceDate(current,Period('-2Y')).strftime("%Y-%m-%d")
    end=y_day
    temp_df=DataAPI.EquDivGet(eventProcessCD = '6',secID=universe1.preview(current) ,beginDate = start ,endDate= end ,field=['secID','perCashDiv','recordDate'],pandas="1").dropna().drop_duplicates('secID',keep='last')
    temp_df.reset_index(inplace=True,drop=True)
    temp_price = DataAPI.MktEqudGet(secID =temp_df.secID , tradeDate= y_day,field=u"secID,closePrice",pandas="1") 

    temp_RPP=DataAPI.FdmtIndiPSPitGet(secID=temp_df.secID,publishDateBegin=start,publishDateEnd=end,field=u"secID,rePS,nAssetPS,publishDate",pandas="1").drop_duplicates('secID')
    temp_RPP.reset_index(drop=True,inplace=True)
    temp_RPP['RPP']=temp_RPP['rePS']/temp_RPP['nAssetPS']
    temp_ind=DataAPI.EquIndustryGet(industry=u'申万',secID=temp_df.secID,field=u"secID,industryName1,isNew",pandas="1")
    temp_ind=temp_ind[temp_ind['isNew']==1]

    temp_df=pd.merge(temp_df,temp_price,on='secID')
    temp_df['bonus_rate']=temp_df['perCashDiv']*100.0/temp_df['closePrice']
    temp_df=pd.merge(temp_df,temp_RPP,on='secID')
    temp_df=pd.merge(temp_df,temp_ind,on='secID')
    return temp_df
```

```python
def ind_rank(temp_df):
    '''
    得到行业排名
    '''
    temp_df_group=temp_df.groupby('industryName1').mean()
    temp_df_group['spread']=temp_df['bonus_rate'].groupby(temp_df['industryName1']).apply(lambda x:x.std()/x.mean())
    temp_df_group['RPP_rank']=temp_df_group['RPP'].rank(ascending=False)
    temp_df_group['spread_rank']=temp_df_group['spread'].rank(ascending=False)
    temp_df_group['rank']=(temp_df_group['RPP_rank']+temp_df_group['spread_rank']).rank()
    temp_df_group.sort_values('rank',inplace=True)
    temp_df_group.reset_index(inplace=True)
    temp_rank=temp_df_group['industryName1']
    return temp_rank
```

# `高息股组合构建`
#### 根据上文一系列验证，本文认为新进行行业优选，再根据股息率进行选股是一个有效的选股思路。
#### 组合回测的具体细节如下：
#### 1. 回测时间：2007年1月至2017年4月：
#### 2. 调仓频率：月度调仓；
#### 3. 买卖价格：月底选股，以下月第一个交易日的均价买卖；
#### 4. 先根据行业平均未分配利润占比和行业股息率spread选择排名靠前的10个行业，形成股票池；
#### 5. 在10个行业的股票池里面，根据股息率从大到小选择股息率最高的30只股票
#### 6. 交易成本：双边千分之三
#### 7. 提出ST，停牌和涨跌停不能买卖的情况。

```python
def indbuy_list(df,current):
    '''
    A股中优选后前10行业中的所有股票
    对所有行业进行等权处理
    对行业内个股进行等权处理
    得到secID及weight
    '''
    current=current
    universe1=DynamicUniverse('A')
    y_day=cal.advanceDate(current,Period('-1B')).strftime("%Y-%m-%d")
    temp_ind=ind_rank(df).head(10)
    temp_df=DataAPI.EquIndustryGet(industry=u'申万',secID=universe1.preview(y_day),field=u"secID,industryName1,isNew",pandas="1")
    temp_df=temp_df[temp_df['isNew']==1]
    temp_weight=temp_df.set_index('industryName1').ix[temp_ind].reset_index().groupby('industryName1').size()
    temp_weight=(temp_weight**(-1.0)*0.1).reset_index()
    temp_weight.columns=['industryName1','weight']
    temp_df=pd.merge(temp_df,temp_weight,on='industryName1')
    temp_df.drop(['industryName1','isNew'],axis=1,inplace=True)
    #temp_df.columns=['q4','q4_weight']
    return temp_df
```

```python
def buy_list(df,df2):
    '''
    分别得到
    1. 前10行业
    2. 所有行业
    3. 后10行业
    股息率前30的股票及其权重
    4. indbuy_list中的股票及其权重
    '''
    temp_buy=pd.DataFrame(columns=['q1','q1_weight','q2','q2_weight','q3','q3_weight'],index=np.arange(0,len(df2)))
    temp_rank=ind_rank(df)
    df.set_index('industryName1',inplace=True)
    temp_1=df.ix[temp_rank.head(10)].sort_values('bonus_rate',ascending=False).head(30).reset_index(drop=True)
    temp_1['weight']=temp_1['bonus_rate']/temp_1['bonus_rate'].sum()
    temp_2=df.sort_values('bonus_rate',ascending=False).head(30).reset_index(drop=True)
    temp_2['weight']=temp_2['bonus_rate']/temp_2['bonus_rate'].sum()
    temp_3=df.ix[temp_rank.tail(10)].sort_values('bonus_rate',ascending=False).head(30).reset_index(drop=True)
    temp_3['weight']=temp_3['bonus_rate']/temp_3['bonus_rate'].sum()
    temp_buy['q1']=temp_1['secID']
    temp_buy['q1_weight']=temp_1['weight']
    temp_buy['q2']=temp_2['secID']
    temp_buy['q2_weight']=temp_2['weight']
    temp_buy['q3']=temp_3['secID']
    temp_buy['q3_weight']=temp_3['weight']
    temp_buy['q4']=df2['secID']
    temp_buy['q4_weight']=df2['weight']
    return temp_buy
```

### 限于内存限制，不能回测更长的区间，简单测了下，超额收益主要集中在14-17年，11-13年跑得和中证500几乎无异。

# 再分别构建4个组合
#### 1. 排名前10行业中股息率最高的30只股票
#### 2. 全行业中股息率最高的30只股票
#### 3. 倒数前10行业中股息率最高的30只股票
#### 4. A股前10行业中所有股票
#### 在此感谢**taotao.li**的[研究 | 因子分析模版, 谢 call 神](https://uqer.datayes.com/community/share/58770bdb89e3ba004defe8e7)

```python
start = '2014-01-31'                       # 回测起始时间
end = '2017-05-19'                         # 回测结束时间
benchmark = 'ZZ500'                        # 策略参考标准
universe = DynamicUniverse('A')            # 证券池，支持股票和基金
freq = 'd'                                 # 策略类型，'d'表示日间策略使用日线回测，'m'表示日内策略使用分钟线回测
refresh_rate = 30 
capital_base = 10000000                     
commission = Commission(0.003,0.003)
# ---------------回测参数部分结束----------------


# 把回测参数封装到 SimulationParameters 中，供 quick_backtest 使用
sim_params = quartz.SimulationParameters(start, end, benchmark, universe, capital_base)
# 获取回测行情数据
idxmap, data = quartz.get_daily_data(sim_params)
# 运行结果
results = {}

# 调整参数，进行快速回测
for i in np.arange(1,5):
     # ---------------策略逻辑部分----------------
    refresh_rate = 30    
    commission = Commission(0.003,0.003)
    def initialize(account):                  # 初始化策略运行环境
        pass

    def handle_data(account):               # 核心策略逻辑    
        #account = context.get_account('fantasy_account')

        current = account.current_date  #定义current 账户的现在时间
        temp_df=buy_list(stock_list(current),indbuy_list(stock_list(current),current))[['q'+str(i),'q'+str(i)+'_weight']].dropna().set_index('q'+str(i))
        buylist=list(temp_df.index)

        #price = context.current_price(tuple(buylist))  #证券参考价，一般使用的是上一日收盘价或者上一分钟收盘价。

        stock_pool = list(set(buylist) & set(account.universe))
        #stock_pool = buylist
        for stk in account.security_position:         #上一K线结束后的有效证券头寸，即持仓数量大于0的证券头寸
            if stk not in buylist:   
                account.order_to(stk, 0)  #对于已经持仓的卖出
            else:
                continue

        for stk in stock_pool:  #买单的证券值，之前已经定义
            account.order_pct_to(stk,len(buylist)/len(stock_pool)*temp_df['q'+str(i)+'_weight'][stk] )  #按权重购买
            
    # ---------------策略逻辑部分结束----------------

    # 把回测逻辑封装到 TradingStrategy 中，供 quick_backtest 使用
    strategy = quartz.TradingStrategy(initialize, handle_data)
    # 回测部分
    bt, acct = quartz.quick_backtest(sim_params, strategy, idxmap, data, refresh_rate=refresh_rate, commission=commission)

    # 对于回测的结果，可以通过 perf_parse 函数计算风险指标
    perf = quartz.perf_parse(bt, acct)

    # 保存运行结果
    tmp = {}
    tmp['bt'] = bt
    tmp['annualized_return'] = perf['annualized_return']
    tmp['volatility'] = perf['volatility']
    tmp['max_drawdown'] = perf['max_drawdown']
    tmp['alpha'] = perf['alpha']
    tmp['beta'] = perf['beta']
    tmp['sharpe'] = perf['sharpe']
    tmp['information_ratio'] = perf['information_ratio']
    
    results[i] = tmp
    print str(i),
print 'done'
```

```python
fig = plt.figure(figsize=(10, 8))
fig.set_tight_layout(True)
ax1 = fig.add_subplot(211)
ax2 = fig.add_subplot(212)
ax1.grid()
ax2.grid()

for qt in results:
    bt = results[qt]['bt']

    data = bt[[u'tradeDate',u'portfolio_value',u'benchmark_return']]
    data['portfolio_return'] = data.portfolio_value/data.portfolio_value.shift(1) - 1.0   # 总头寸每日回报率
    data['portfolio_return'].ix[0] = data['portfolio_value'].ix[0] / 10000000.0 - 1.0
    data['excess_return'] = data.portfolio_return - data.benchmark_return                 # 总头寸每日超额回报率
    data['excess'] = data.excess_return + 1.0
    data['excess'] = data.excess.cumprod()                # 总头寸对冲指数后的净值序列
    data['portfolio'] = data.portfolio_return + 1.0     
    data['portfolio'] = data.portfolio.cumprod()          # 总头寸不对冲时的净值序列
    data['benchmark'] = data.benchmark_return + 1.0
    data['benchmark'] = data.benchmark.cumprod()          # benchmark的净值序列
    results[qt]['hedged_max_drawdown'] = max([1 - v/max(1, max(data['excess'][:i+1])) for i,v in enumerate(data['excess'])])  # 对冲后净值最大回撤
    results[qt]['hedged_volatility'] = np.std(data['excess_return'])*np.sqrt(252)       #波动率
    results[qt]['hedged_annualized_return'] = (data['excess'].values[-1])**(252.0/len(data['excess'])) - 1.0            #年华收益率
    ax1.plot(data['tradeDate'], data[['portfolio']], label=str(qt))
    ax2.plot(data['tradeDate'], data[['excess']], label=str(qt))


ax1.legend(loc=0, fontsize=12)
ax2.legend(loc=0, fontsize=12)
ax1.set_ylabel(u"净值", fontproperties=font, fontsize=16)
ax2.set_ylabel(u"对冲净值", fontproperties=font, fontsize=16)
ax1.set_title(u"不同选股组合净值走势", fontproperties=font, fontsize=16)
ax2.set_title(u"不同选股组合对冲中证500指数后净值走势", fontproperties=font, fontsize=16)
```
