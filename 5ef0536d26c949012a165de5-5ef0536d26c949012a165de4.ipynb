# 折价基金轮动

```python
from CAL.PyCAL import *
import numpy as np
import pandas as pd
cal= Calendar('China.SSE')
pd.set_option('display.height',1000)
pd.set_option('display.max_rows',500)
pd.set_option('display.max_columns',500)
pd.set_option('display.width',1000)
def get_tickers_close(trade_date):
    if cal.isHoliday(trade_date):
        trade_date= cal.advanceDate(trade_date, '1B', BizDayConvention.Following)
        
    tickers_data=DataAPI.MktFunddGet(secID=u"",ticker=u"",tradeDate=trade_date,beginDate="",endDate="",field=u"ticker,secID,discountRatio,turnoverValue,secShortName",pandas="1")
    filterData= tickers_data[(tickers_data.discountRatio > 0.03) & (tickers_data.turnoverValue> 500000.0)]
    if filterData.empty:
        return
    secIDs_list = filterData['secID'].tolist()
    secIDs_str = ",".join(secIDs_list);
    
    tickers_data = DataAPI.FundGet(secID=secIDs_str,ticker=u"",etfLof=u"",listStatusCd=u"",category=['E','H'],\
    idxID=u"",idxTicker=u"",operationMode=u"",beginDate="",endDate="",status="", \
                                   field=u"ticker,secID,idxTicker,secShortName,category,isQdii,isClass,status",pandas="1")
   #  过滤 分级基金
   # tickers_data.drop_duplicates(subset=['ticker'], keep='first', inplace=True)
   # print (tickers_data[(tickers_data.isClass == 0) & (tickers_data.isQdii == 0)])
    secIDs_list = (tickers_data[(tickers_data.isClass == 0) & (tickers_data.isQdii == 0)])['secID'].tolist() 
    return  secIDs_list
def  get_tickers_close_pool(begin_date,end_date):
    begin = Date.strptime(begin_date, '%Y%m%d')
    end = Date.strptime(end_date, '%Y%m%d')
    span = Period("1M");
    begin_str = begin.strftime("%Y%m%d");
    nextDate = begin + span
    secID_list = get_tickers_close(begin_str);
    while nextDate < end :
        nextDate_str = nextDate.strftime('%Y%m%d');
        secID_list = list(set(secID_list + get_tickers_close(nextDate_str)))
        begin = nextDate
        nextDate = begin + span
    return secID_list
print  get_tickers_close('20180104')
```
