# **一.传统均线策略**
- ### 传统均线策略大家都非常熟悉，最简单的就是移动平均线，即股价前几日的均值。(MACD详情请见[『研究｜技术指标｜第一弹』MACD](https://uqer.datayes.com/community/share/5799b908228e5ba291060674))
- ### 我们可以便捷的使用快速回测框架来看看不同均线组合下，传统均线策略的表现情况。

```python
#加载需要用到的包
import numpy as np
import pandas as pd
from CAL.PyCAL import *    # CAL.PyCAL中包含font
font.set_size(22)
small_font = font.copy()
small_font.set_size(12)

import matplotlib.pyplot as plt
```

- ### 我们对常见的周期5，10，20，30，60，120，240两两组合

```python
Tuple =[]  #配对的组合
a = [5,10,20,30,60,120,240]   
for i in range(len(a)-1):
    for j in range(len(a)-1-i):
        Tuple.append((a[i],a[j+i+1]))   #配对
print Tuple
```

- ### 研究时间：20140801-20160801
- ### 策略逻辑：金叉买入(快线上穿慢线)；死叉卖出(快线下穿慢线)

```python
#回测代码
# 可编辑部分与 strategy 模式一样，其余部分按本例代码编写即可

# -----------回测参数部分开始，可编辑------------

start = '2014-08-01'
end = '2016-08-01'
benchmark = 'HS300'                        # 策略参考标准
universe = set_universe('HS300','20140801')  # 证券池，支持股票和基金
capital_base = 1000000

# ---------------回测参数部分结束----------------

# 把回测参数封装到 SimulationParameters 中，供 quick_backtest 使用
sim_params = quartz.SimulationParameters(start, end, benchmark, universe, capital_base)
# 获取回测行情数据
idxmap, data = quartz.get_daily_data(sim_params)
# 运行结果
results = {}

# 调整参数快速回测
for x in range(len(Tuple)):
    

    refresh_rate = 1
    
    k1 = Tuple[x][0]
    k2 = Tuple[x][1]
    
    def initialize(account):
        account.sig={}  #记录持有还是卖出信号
        for i in universe:
            account.sig[i]=0
        pass
    def handle_data(account):
        price = account.get_attribute_history('closePrice',k2)
        buy=[]
        sell=[]
        for stk in account.universe:
            
            MAk1 = price[stk][-k1:].mean()
            MAk2 = price[stk][-k2:].mean()
            
            if MAk1 >= MAk2:           
                if account.sig[stk] == 0:   #金叉买入
                    buy.append(stk)
                account.sig[stk]=1
            elif MAk1 < MAk2:
                if account.sig[stk] == 1:   #死叉卖出
                    sell.append(stk)
                account.sig[stk]=0
        if len(sell)>0:
            for j in sell:
                order_to(j,0)
        if len(buy)>0:
            for i in buy:
                order_pct_to(i, 1.0/len(buy))
       
        # 调用 account.my_parameter 进行计算或其他操作
        return

    # ---------------策略逻辑部分结束----------------

    # 把回测逻辑封装到 TradingStrategy 中，供 quick_backtest 使用
    strategy = quartz.TradingStrategy(initialize, handle_data)
    # 回测部分
    bt, acct = quartz.quick_backtest(sim_params, strategy, idxmap, data, refresh_rate = refresh_rate)

    # 对于回测的结果，可以通过 perf_parse 函数计算风险指标
    perf = quartz.perf_parse(bt, acct)

    # 保存运行结果
    results[x] = perf
    
```

```python
#得到各个组合的风险指标
result = pd.DataFrame()
result['均线组合(快，慢)'] = Tuple
result['年化收益'] = [results[i]['annualized_return'] for i in range(len(Tuple))]
result['最大回撤'] = [results[i]['max_drawdown'] for i in range(len(Tuple))]
result['收益波动率'] = [results[i]['volatility'] for i in range(len(Tuple))]
result['信息比率'] = [results[i]['information_ratio'] for i in range(len(Tuple))]
```

```python
#画图
fig = plt.figure(figsize = (16,6))
ax = fig.add_subplot(111)
result.set_index('均线组合(快，慢)').plot(kind='bar',figsize=(14, 7),ax=ax,color=['r','y','#DB7093','b'],title=u'传统均线组合表现')
ax.legend(map(lambda x: x.decode('utf-8'),result.columns.values[1:]),prop=small_font,loc='best')
ax.set_xlabel(u'均线组合(快，慢)', fontproperties=small_font)
plt.xticks(fontproperties = small_font, rotation = 70)
ax.title.set_font_properties(font)
ax.set_axis_bgcolor('white')
ax.grid(color='gray', alpha=1.0, axis='y')
```

- ### 为什么要这么改进呢？研报里从期望的角度去解释了一下，你们感受一下：
>  ## 凸组合定义：
> - ### 若x满足$x=w_1x_1+w_2x_2+\cdots+w_nx_n$，其中$w_i>0,w_1+w_2+\cdots+w_n=1$,则x是$x_1,x_2,,,x_n$的凸组合
> - ### 从形式上看，凸组合可以看成期望（其实并不是）
> - ### 当前价格$p_{t+\epsilon}$与开仓价格$p_t$的凸组合当作价格的期望。。。
> - ### $p^*_{t+\epsilon}=P[S_{t+\epsilon}=1]p_t+(1-P[S_{t+\epsilon}=1])p_{t+\epsilon}$,其中$S_t=1$表示快线大于慢线

- ### 最后因为概率无法估计，就以金叉时的价格作为平仓的标准。。。
- ### 好吧，我是没看懂，完全跟期望不是一回事嘛。感觉就是动量，他说要用凸组合，最后也没用上，就用了金叉时的价格
- ### 不管怎么想的，先看看效果怎么样吧。
- ### 回测的时间和股票范围仍然是之前的,策略逻辑则改成刚刚所描述的那样：

```python
#回测代码
# 可编辑部分与 strategy 模式一样，其余部分按本例代码编写即可

# -----------回测参数部分开始，可编辑------------

start = '2014-08-01'
end = '2016-08-01'
benchmark = 'HS300'                        # 策略参考标准
universe = set_universe('HS300','20140801')  # 证券池，支持股票和基金
capital_base = 1000000

# ---------------回测参数部分结束----------------

# 把回测参数封装到 SimulationParameters 中，供 quick_backtest 使用
sim_params = quartz.SimulationParameters(start, end, benchmark, universe, capital_base)
# 获取回测行情数据
idxmap, data = quartz.get_daily_data(sim_params)
# 运行结果
results_2 = {}

# 调整参数快速回测
for x in range(len(Tuple)):
    

    refresh_rate = 1
    
    k1 = Tuple[x][0]
    k2 = Tuple[x][1]
    
    def initialize(account):
        account.cost_price = {}   #成本价
        account.sig={}     #1为持有信号，0为不持有信号
        for t in universe:
            account.cost_price[t] = 0
            account.sig[t] = 0
        pass
    def handle_data(account):
        price = account.get_attribute_history('closePrice',k2)
        buy=[]
        sell=[]
        for stk in account.universe:
            MAk1 = price[stk][-k1:].mean()
            MAk2 = price[stk][-k2:].mean()
  
            if MAk1 >= MAk2:
                if account.sig[stk] == 0:  #金叉买入
                    buy.append(stk)
                    account.cost_price[stk] = price[stk][-1]
                account.sig[stk]=1
            elif MAk1 < MAk2:
                if account.sig[stk] == 1:  #死叉
                    if price[stk][-1] < account.cost_price[stk]:   #价格小于成本价，卖出
                        sell.append(stk)
                        account.cost_price[stk] = 0
                    elif price[stk][-1] >= account.cost_price[stk]:  #价格大于成本价，买入
                        buy.append(stk)
                account.sig[stk]=0
        if len(sell)>0:
            for j in sell:
                order_to(j,0)
        if len(buy)>0:
            for i in buy:
                order_pct_to(i, 1.0/len(buy))
        
        return

    # ---------------策略逻辑部分结束----------------

    # 把回测逻辑封装到 TradingStrategy 中，供 quick_backtest 使用
    strategy = quartz.TradingStrategy(initialize, handle_data)
    # 回测部分
    bt, acct = quartz.quick_backtest(sim_params, strategy, idxmap, data, refresh_rate = refresh_rate)

    # 对于回测的结果，可以通过 perf_parse 函数计算风险指标
    perf = quartz.perf_parse(bt, acct)

    # 保存运行结果
    results_2[x] = perf
    
```

- ### 既然他说要用凸组合，我就用凸组合试着优化一下均线策略吧。凸组合的本质其实就是加权平均，不妨就简单从这个角度去优化一下。
- ### 计算成本价格时，为了避免单日价格的随机波动，可以取前几日的加权平均价格（以距离为权重），取多久的呢，我们又用快速回测框架来看看。

```python
formerdays=np.arange(1,30) #取前n日的价格算加权平均
```

```python
#回测代码
# 可编辑部分与 strategy 模式一样，其余部分按本例代码编写即可

# -----------回测参数部分开始，可编辑------------

start = '2014-08-01'
end = '2016-08-01'
benchmark = 'HS300'                        # 策略参考标准
universe = set_universe('HS300','20140801')  # 证券池，支持股票和基金
capital_base = 1000000

# ---------------回测参数部分结束----------------

# 把回测参数封装到 SimulationParameters 中，供 quick_backtest 使用
sim_params = quartz.SimulationParameters(start, end, benchmark, universe, capital_base)
# 获取回测行情数据
idxmap, data = quartz.get_daily_data(sim_params)
# 运行结果
results_3 = {}

# 调整参数快速回测
for x in range(len(formerdays)):
    

    refresh_rate = 1
    
    k1 = 5
    k2 = 20
    N = formerdays[x]
    
    def initialize(account):
        account.cost_price = {}   #成本价
        account.sig={}     #1为持有信号，0为不持有信号
        for t in universe:
            account.cost_price[t] = 0
            account.sig[t] = 0
        pass
    def handle_data(account):
        price = account.get_attribute_history('closePrice',32)
        buy=[]
        sell=[]
        for stk in account.universe:
            MAk1 = price[stk][-k1:].mean()
            MAk2 = price[stk][-k2:].mean()
  
            if MAk1 >= MAk2:
                if account.sig[stk] == 0:  #金叉买入
                    buy.append(stk)
                    dictance=0
                    cost=0
                    for i in range(N):
                        dictance += (N-i)
                    for i in range(N):
                        cost += (N-i)*1.0/dictance*price[stk][-(i+1)]
                    account.cost_price[stk] = cost
                account.sig[stk]=1
            elif MAk1 < MAk2:
                if account.sig[stk] == 1:  #死叉
                    if price[stk][-1] < account.cost_price[stk]:   #价格小于成本价，卖出
                        sell.append(stk)
                        account.cost_price[stk] = 0
                    elif price[stk][-1] >= account.cost_price[stk]:  #价格大于成本价，买入
                        buy.append(stk)
                account.sig[stk]=0
        if len(sell)>0:
            for j in sell:
                order_to(j,0)
        if len(buy)>0:
            for i in buy:
                order_pct_to(i, 1.0/len(buy))
        
        return

    # ---------------策略逻辑部分结束----------------

    # 把回测逻辑封装到 TradingStrategy 中，供 quick_backtest 使用
    strategy = quartz.TradingStrategy(initialize, handle_data)
    # 回测部分
    bt, acct = quartz.quick_backtest(sim_params, strategy, idxmap, data, refresh_rate = refresh_rate)

    # 对于回测的结果，可以通过 perf_parse 函数计算风险指标
    perf = quartz.perf_parse(bt, acct)

    # 保存运行结果
    results_3[x] = perf
    
```

```python
#得到各个组合的风险指标
result3 = pd.DataFrame()
result3['前N日加权平均'] = formerdays
result3['年化收益'] = [results_3[i]['annualized_return'] for i in range(len(formerdays))]
result3['最大回撤'] = [results_3[i]['max_drawdown'] for i in range(len(formerdays))]
result3['收益波动率'] = [results_3[i]['volatility'] for i in range(len(formerdays))]
result3['信息比率'] = [results_3[i]['information_ratio'] for i in range(len(formerdays))]
```

```python
#画图
fig = plt.figure(figsize = (16,6))
ax = fig.add_subplot(111)
result3.set_index('前N日加权平均').plot(kind='bar',figsize=(14, 7),ax=ax,color=['r','y','#DB7093','b'],title=u'凸组合改进均线表现')
plt.legend(map(lambda x: x.decode('utf-8'),result2.columns.values[1:]),prop=small_font,loc='best')
ax.set_xlabel(u'前N日加权平均)', fontproperties=small_font)
plt.xticks(fontproperties = small_font, rotation = 70)
ax.title.set_font_properties(font)
ax.set_axis_bgcolor('white')
plt.grid(color='gray', alpha=1.0, axis='y')
```
