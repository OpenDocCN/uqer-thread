    作为研报的超级超级简化版，总结下，这个策略寻找alpha失败，我觉得可能有几点原因：
    1.因子处理，这步工作量有点大，我觉得这种 在线平台处理数据还是好吃力
    2.最重要的是f，也就是回归的预测，我调试过程中看了看我回归的R^2，对hs300，在0.3左右，而研报说他的结果0.4几，什么概念，可以理解为相关系数》0.6了，我的结果是随着股票数量增加R^2骤减，有待研究。其实个人觉得涉及markowitz，必然出现预测难的问题，是真的有点不相信这个了，你要是预测能准，直接选股吧，还对冲啥，哈哈，开玩笑，减小风险嘛。
    3.为了解决预测难问题，文中用了几种方案，其中我试了加权回归，其实回归系数和R^2影响很小，所以也没有用，另外在协方差估计中，作者也用了一些方法由日频率转换月频率，我则直接用的月频率，在目标函数中，我把风险项系数设为0，调参结果也不对，所以协方差这一块也没深入。如果大体结果不错的话，这些细节可以进一步优化的。
    4.研报上ms以全A来求解的，我试了下，算不出来太慢了，最后还是选的HS300求解，这里bench也是HS300,所以h和hbench一样（但是w不同哦）。
