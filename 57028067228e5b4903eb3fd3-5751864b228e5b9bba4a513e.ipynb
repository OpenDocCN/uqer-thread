# 程序员学量化投资（六）：成交量之缩量增量

最近在学习量化选股，趋势追踪模型中 成交量与价格 是在股票走势中相当重要的因素。这一期，我们先熟悉下成交量，通过程序判断成交量的缩量和增量出现的时间。

成交量，股票有人卖有人买就是一次成交，而卖者是看空后市，买者是看涨后市。所以成交量本质上市场交易者的分歧。成交量越大分歧越大，后市反转的可能性越大；成交量越小，分歧越小，后市走势越明朗。
所以我们这段程序，简单的判断增量和缩量发生的时间。

我们用图表的方式更直观的了解 成交量的走势关系，慢线较平稳，可以反映长期的平均成交量并作为成交量增减的对比数据。而5日短线则较好的反映出近期的增量区间和缩量区间。

```python
#构建DataFrame
df = DataFrame(voldata)
pd.set_option('display.width', 400)
df2=df.set_index('time')
df2['currentVOL'].plot(kind='bar',color='b')
df2['5daysVOL'].plot(color='r')
df2['15daysVOL'].plot(color='g')
plt.legend(loc='best')
```
