# 会计估值视角下的价值因子和size-effect探讨
## 1.模型：

在合理的会计准则下，假设财务报表中的数据满足清洁盈余关系(Clean Surplus Relation,CSR)，我们有
$$B_{t+1}=B_{t}-d_{t+1}+Earnings_{t+1}$$ 
其中$B_{t}$是t期的账面价值, $d_{t+1}$是t+1期的payout（现金分红+回购股份),$\mathrm{Earnings}_{t+1}$ 是t+1期的利润。
上式经过变形,同时取数学期望（严格地说是在t期的条件期望）可得：

$$E(\frac{P_{t+1}+d_{t+1}-P_{t}}{P_{t}})=E(\frac{Earnings_{t+1}}{P_t})+E(\frac{P_{t+1}-B_{t+1}-(P_t-B_t)}{P_t})$$
等式的左边是t+1期的预期收益率，右边第一项是未来t+1期的预期earning yield，第二项和预期的premium, $P_t-B_t$ 变化有关。
所以预测未来一期的收益率变成了预测未来一期的利润以及premium。而t+1期的premium主要取决于t+1期后的成长性，成长性越高第二项对于预期收益率的贡献就越大。

总之，在以上分析框架下我们有以下结论：
1. 股票收益率取决于短期利润和长期成长性;
2. 成长性越高，短期利润的影响就越小。

所以我们可以将预测股票收益率的任务分解为：
1. 预测短期利润；
2. 预测长期的成长性带来的premium的增加，

对于短期利润，很自然想到用分析师的一致预期数据。但因为系统性高估和数据覆盖率不全的问题，这里采用t期的净利润来作为估计值。即$\frac{E_t}{P_t}$。
对于premium的增加，观察到到第二项中包含了$\frac{B_t}{P_t}$，而且该项也确实是代表成长性的指标，启发我们在实证的时候将这一项作为预期收益率的解释变量。

所以有收益率的预测模型：
$$R_{t+1}= \omega_1 \frac{E_t}{P_t}+ \omega_2 \frac{B_t}{P_t}+\epsilon_t$$

从信号合成的角度，上式将两个信号加权合成复合信号，信号合成的方法一般有等权，最大化IC，最大化IR几种。但由于“成长性越高，短期利润对股票收益率的影响就越小。”，对于整体股票空间采取均一的权重会忽视权重对于成长性的依赖。而市值规模可以作为衡量成长性高低的一个简单指标，市值越小成长性越高，收入的不确定性也增加，所以这里将同一时刻的全部股票按照市值大小分成10组，每一组内部分别作线性回归计算权重。同时为了避免异常值对结果的影响，这里采用Theil-Sen 回归， 这是一种非参数统计方法，利用计算机重复取样，分别回归估计参数，然后得到参数的分布，最后根据分布获得一组稳健的参数估计值。

## 3.结果讨论
表格中size分组的编号从0-9代表市值从小到大。
从因子的分布上看，随着市值分组由小到大：

1. BP的组内均值的历史平均逐步提升，有比较好的单调性，BP的组内中位数的历史平均虽然也有上升趋势但单调性较差，这与美股中的结果是完全反向的。

2. 同样地，EP的组内均值和组内中位数的历史平均逐步提升，有比较好的单调性。
以上规律符合大家“市值越大，估值越低”的印象。

对于BP EP 两个因子的权重而言，随着市值分组由小到大：
1. BP因子的权重逐步减小系数的显著性也越来越小，对于市值最大三个分组，即市场大约前800只股票，BP对收益率没有预测效果。
2. EP因子的权重先逐步增大，但对于最大的几组权重有所下降，系数的显著性的变化趋势则不太明显。比较特殊的是市值最大的分组，也就是市场大约前250只股票，BP对收益率的预测是负向的，可能是样本点比较少的原因。

以上现象说明，随着市值增大，预测收益率的权重从BP逐步向EP倾斜。



根据各组的历史BP/EP均值，我们可以估计该均值水平下的收益率，然后对历史上的估计收益率取平均，得到各组的fitted_ret_by_grp_mean 数据。可以看到市值越小该平均收益率越大，单调性相当好。用类似的方法计算fitted_ret_by_grp_median，也可以得出类似的结论。

这个和三因子模型里的size-effect是一致的，即SMB预期有正收益。模型并中没有直接引入size变量，只要满足因子权重随着size增加预测收益率的权重从BP逐步向EP倾斜，就可以衍生出size-effect。推而广之，还能找到其他的一些类似size的特征，价值因子加权的权重和特征自身的水平相关，也可以产生xx-effect。

```python

from sklearn.linear_model import LinearRegression, TheilSenRegressor
import pandas as pd
from sklearn.linear_model import RANSACRegressor
import numpy as np
from scipy import stats


def cross_sectional_regress_by_size_grp(grp_df, estimator):
	estimator.fit(grp_df[['BP', 'EP']], grp_df['year_return'])
	mean_value = grp_df[['BP','EP']].mean()
	median_value = grp_df[['BP','EP']].median()

	# mean =grp_df[['BP', 'EP']].mean()
	forecast_ret_mean = estimator.predict([mean_value])
	forecast_ret_median = estimator.predict([median_value])

	record = {"BP_mean": mean_value['BP'],
	          "EP_mean": mean_value['EP'],
	          "BP_median": median_value['BP'],
	          "EP_median": median_value['EP'],
	          "BP_coef": estimator.coef_[0],
	          "EP_coef": estimator.coef_[1],
	          "fitted_ret_by_grp_mean": forecast_ret_mean[0],
	          "fitted_ret_by_grp_median": forecast_ret_median[0]
	          }
	index = ['BP_coef', 'BP_mean', 'BP_median', 'EP_coef', 'EP_mean', 'EP_median', 'fitted_ret_by_grp_mean', 'fitted_ret_by_grp_median']

	res = pd.Series(record, index=index)


	return res



def regress_on_bp(grp_df,estimator):
	estimator.fit(grp_df[['BP']], grp_df['year_return'])
	return estimator.coef_

def regress_on_ep(grp_df,estimator):
	estimator.fit(grp_df[['EP']], grp_df['year_return'])
	return estimator.coef_

estimators = [('OLS', LinearRegression()),
              ('Theil-Sen', TheilSenRegressor(random_state=42)),
              ('RANSAC', RANSACRegressor(random_state=42)), ]
ols  = LinearRegression()
ts_regressor = TheilSenRegressor(random_state=243,max_iter=500,tol=0.0001,n_subsamples=100,fit_intercept=False)

value_factor_trading_days = DataAPI.TradeCalGet(exchangeCD=u"XSHG,XSHE", beginDate=u"20070101", endDate=u"20161231", field=u"", pandas="1")
td_year_end = value_factor_trading_days.query("isYearEnd==1 and exchangeCD=='XSHG'")
common_key = ["secID", "publishDate", "endDate", "reportType"]
common_key1 = ["secID", "publishDate", "endDate",]

factor_field = ['PB','PE','LCAP']
# value_factor_date = '20151231'
# c2c_return_year = int(value_factor_date[:4]) - 1
regress_res_by_size_grp_list = []

for value_factor_date in td_year_end['calendarDate']:
	value_factor_date = ''.join(value_factor_date.split('-'))
	c2c_return_year = int(value_factor_date[:4])+1
	print c2c_return_year
	value_data = DataAPI.MktStockFactorsOneDayGet(tradeDate=value_factor_date, secID=u"", ticker=u"", field=common_key1 + factor_field, pandas="1")
	value_data.loc[value_data['PB']==0,'PB'] = np.nan
	value_data.loc[value_data['PE']==0,"PE"] = np.nan

	value_data['BP'] = 1./value_data['PB']
	value_data['EP'] = 1./value_data['PE']

	# todo value growth
	# oper_profit = DataAPI.FdmtISGet(ticker=u"",secID=u"",reportType=u"",endDate=u"20150105",beginDate=u"20180629",publishDateEnd=u"",
	#                                 publishDateBegin=u"",endDateRep="",beginDateRep="",beginYear="",endYear="",fiscalPeriod="",
	#                                 field=common_key+['operateProfit'],pandas="1")

	field = ['secID','monthBeginDate','endDate','openPrice','closePrice','return']

	dec_reinv_ret = DataAPI.MktEqumAdjAfGet(secID=u"",ticker=u"",monthEndDate=u"",beginDate=str(c2c_return_year)+"1201",endDate=str(c2c_return_year)+"1231",isOpen=u"",field=field,pandas="1")

	predec_reinv_ret = DataAPI.MktEqumAdjAfGet(secID=u"",ticker=u"",monthEndDate=u"",beginDate=str(c2c_return_year-1)+"1201",endDate=str(c2c_return_year-1)+"1231",isOpen=u"",field=field,pandas="1")

	df_merge = dec_reinv_ret.merge(predec_reinv_ret,left_on='secID',right_on='secID',suffixes=('_0', '_m1'))
	df_merge['year_return'] = (df_merge['closePrice_0']-df_merge['closePrice_m1'])/df_merge['closePrice_m1']
	df_merge = df_merge.merge(value_data, on='secID')
	df_merge.sort_values(by='LCAP',ascending=True,inplace=True)
	# todo : add earnings
	df_merge = df_merge[~df_merge[['BP', 'EP', 'LCAP']].isnull().any(axis=1)]
	grp = pd.qcut(df_merge['LCAP'],10,range(10))
	df_merge['size_grp'] = grp
	regress_res_by_size_grp = df_merge.groupby('size_grp').apply(cross_sectional_regress_by_size_grp, ts_regressor)

	regress_res_by_size_grp['forward_year']=c2c_return_year
	regress_res_by_size_grp_list.append(regress_res_by_size_grp)

regress_res_by_size_grp = pd.concat(regress_res_by_size_grp_list, axis=0)
# coef_bp = df_merge.groupby('size_grp').apply(regress_on_bp,ts_regressor)
# print coef_bp
# coef_ep = df_merge.groupby('size_grp').apply(regress_on_ep,ts_regressor)
# print coef_ep
regress_res_by_size_grp.reset_index(drop=False, inplace=True)

def cal_stat_by_size_group(df):
	index = ['BP_coef', 'BP_mean', 'BP_median', 'EP_coef', 'EP_mean', 'EP_median', 'fitted_ret_by_grp_mean', 'fitted_ret_by_grp_median']

	bp_ttest_res = stats.ttest_1samp(df['BP_coef'], 0.0)
	ep_ttest_res = stats.ttest_1samp(df['EP_coef'], 0.0)
	mean = df[index].mean(axis=0)
	# forecast_ret = estimator.predict(mean)
	record = mean
	record = record.append(pd.Series({"BP_coef_tstat": bp_ttest_res[0],
	                        "EP_coef_tstat": ep_ttest_res[0]})
	              )
	#           "fitted_return": forecast_ret[0]}
	# res = pd.Series(record, index=["BP_coef", "EP_coef", "BP_mean", "EP_mean", "fitted_return"])
	return record


table_6_res = regress_res_by_size_grp.groupby(['size_grp']).apply(cal_stat_by_size_group)

pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
pd.set_option('precision', 3)
index = [
 'EP_coef',
 'EP_coef_tstat',
 'BP_coef',
 'BP_coef_tstat',
 'BP_mean',
 'BP_median',
 'EP_mean',
 'EP_median',
 'fitted_ret_by_grp_mean',
 'fitted_ret_by_grp_median']
print table_6_res.T.reindex(index).to_html()
```
