# **`5217策略`**
***
#### 策略思想来自于雪球用户:5217指数.
## **策略逻辑：**
- #### 每年5月，10月第一个交易日调仓
- #### 每次选取连续5季，营业收入和净利润增长率均大于17%的5217股（增长率使用同比）
- #### 财务数据选取累计的，也就是3月是一个季度的，6月是两个季度的，9月是三个季度的，12月是四个季度的
- #### 每次先卖出所有股票，再等权买入所选股票

## **策略实现：**

```python
#加载需要用的包
import numpy as np
import pandas as pd
```

### **处理日期**
- #### 选取出每年4月和9月底最后一个交易日，存在refresh_date里

```python
#处理日期，先选取出所有每年月末交易日
data = DataAPI.TradeCalGet(exchangeCD=u"XSHG",beginDate='20050101',endDate='20160101',field=['calendarDate','isWeekEnd','isMonthEnd'],pandas="1")
data = data[data['isMonthEnd'] == 1]
date_list = map(lambda x: x[0:4]+x[5:7]+x[8:10], data['calendarDate'].values.tolist())
```

```python
#处理每月末交易日，选取出4月和9月最后一个交易日
refresh_date=np.zeros(len(date_list)/6)   #一年只有1/6个月会调仓
for i in range(1,len(date_list)+1):      #以index除以12的余数来判断当前index所对应的是几月
    if i%12 == 4:                 #余数为4，表示当前index对应的月是4月，把它存起来，放进refresh_date里，下面一样
        refresh_date[int(i/12)*2+0] = date_list[i-1]
    if i%12 == 10:
        refresh_date[int(i/12)*2+1] = date_list[i-1]
refresh_date=list(refresh_date)        #把refresh_date存为list
refresh_date=[str(int(x)) for x in refresh_date]  
refresh_date=refresh_date+['20160429']   #由于研究时间到了20160401，所以需要把这个数据加进去
refresh_date.sort(reverse=False)  
```

### **计算某只股票某个日期前的连续5季的营业收入和净利润同比增长率序列**
- #### 我们不妨定义一个函数来处理这个问题。
- #### 先看一下获取财报的API调出来的数据是怎样的：

```python
#选取财报，具体的参数请参考https://uqer.datayes.com/data/search/%E5%90%88%E5%B9%B6
DataAPI.FdmtISGet(ticker=u"000002",publishDateEnd='20160803',field=u"secID,publishDate,fiscalPeriod,endDate,endDateRep,revenue,NIncomeAttrP").head(10)
```

- #### 可以看到对于某一年的财报数据而言，有两个值，其中一个是第二年公布的类似于财报重述，我们可以选取调仓日能获取的最新的信息
- #### 这里需要注意的细节很多，还有会计长度的问题，这里可以参考魏老师的帖子(https://uqer.datayes.com/community/share/56fd119d228e5b1f8a1a8fb0)
- #### 下面就可以写这个函数了，传入股票代码和日期，就可以得到当前日期前5个规定财报季的营业收入和净利润同比增长率数据

```python
#取出财报四季末日期
tmp = DataAPI.FdmtISGet(ticker=u"000002",publishDateEnd='20160803',field=u"endDate")
fiscaldate = tmp.groupby('endDate').sum()
fiscaldate = list(fiscaldate.index)
fiscaldate = map(lambda x: x[0:4]+x[5:7]+x[8:10], fiscaldate)
print fiscaldate
```

- #### 这里需要避免使用未来函数。
- #### 举个例子，2014年第一季度的财务数据在2015年4月左右会有一个更新的数据，但是如果此时我们在2014年5月调仓时就不能使用2015年的那个数据；如果我们在2015年5月调仓就可以使用这个数据。

```python
#获取财务数据，这里是调仓日前的那个财季能获得的最新数据，避免未来函数
def get_fiscal1(stk,date):
    data = DataAPI.FdmtISGet(secID=stk,endDate=date,beginDate=date,reportType='Q1,S1,CQ3,A',field=u"revenue,NIncomeAttrP") 
    return data.iat[len(data)-1,0],data.iat[len(data)-1,1]
```

```python
#获取财务数据，这里是调仓日能获取的之前财季的最新数据
def get_fiscal2(stk,date):
    data = DataAPI.FdmtISGet(secID=stk,endDate=date,beginDate=date,reportType='Q1,S1,CQ3,A',field=u"revenue,NIncomeAttrP") 
    return data.iat[0,0],data.iat[0,1]
```

```python
#计算某只股票某个日期前的连续5季的营业收入和净利润同比增长率序列
def get_increase(stk,date):  #输入股票和日期即可
    increase = pd.DataFrame()         #用于存增长率数据
    increase['revenue'] = np.zeros(5)     
    increase['NIncome'] = np.zeros(5)
    if date[5] == '5':
        date = str((int(date[:4]))*10000+331)
    if date[5] == '1':
        date = str((int(date[:4]))*10000+930)
    index = fiscaldate.index(date)
    for i in range(5):
        try:
            if i==0:             
                revenue1,NIncome1 = get_fiscal1(stk,fiscaldate[index-i])  
                revenue2,NIncome2 = get_fiscal2(stk,fiscaldate[index-i-4])
                if NIncome2<0 and NIncome1>0:
                    increase.iat[i,0] = revenue1*1.0/revenue2-1    #计算营业收入同比增长率
                    increase.iat[i,1] = 0.2    #计算净利润同比增长率
                if NIncome1>0 and NIncome2>0:
                    increase.iat[i,0] = revenue1*1.0/revenue2-1    #计算营业收入同比增长率
                    increase.iat[i,1] = NIncome1*1.0/NIncome2-1    #计算净利润同比增长率
            else:
                revenue1,NIncome1 = get_fiscal2(stk,fiscaldate[index-i])
                revenue2,NIncome2 = get_fiscal2(stk,fiscaldate[index-i-4])
                if NIncome2<0 and NIncome1>0:
                    increase.iat[i,0] = revenue1*1.0/revenue2-1    #计算营业收入同比增长率
                    increase.iat[i,1] = 0.2    #计算净利润同比增长率
                if NIncome1>0 and NIncome2>0:
                    increase.iat[i,0] = revenue1*1.0/revenue2-1    #计算营业收入同比增长率
                    increase.iat[i,1] = NIncome1*1.0/NIncome2-1    #计算净利润同比增长率
        except:
            continue
    return increase                 #返回连续5季的数据
```

- #### 先以2016年5月1日为例，即2016年第一个财报季：

```python
stocks = []
for stk in universe:
    increase = get_increase(stk,'20160501')   #获取当前日期前5个规定财报季的营业收入和净利润同比增长率的数据
    increase = increase[(increase['revenue']>0.17) & (increase['NIncome']>0.17)]   #选取营业收入增长率均大于17%的和净利润增长率均大于17%的
    if len(increase) == 5:
        stocks.append(stk)
```

- #### 获取每个调仓日全Ａ股的5217股票

```python
pick_data = {}   #用于存取每次调仓选取出的股票，key为调仓日期
for date in refresh_date:
    stocks = []
    date = str((int(date[:6])+1)*100+1)
    for stk in universe:
        increase = get_increase(stk,date)   #获取当前日期前5个规定财报季的营业收入和净利润同比增长率的数据
        increase = increase[(increase['revenue']>0.17) & (increase['NIncome']>0.17)]   #选取营业收入增长率均大于17%的和净利润增长率均大于17%的
        if len(increase) == 5:
            stocks.append(stk)
    pick_data[date] = stocks
```
