# 股本加报率估值算法

```python
# coding:utf-8
# ========================================================
# 功能：股本回报率估值算法
# 日期：2017-06-07
# 原理：根据给定的期望收益率、投资年数和当前的EPS，通过数据统计得出平均市盈率和盈利增长率推算出最后的股价再折现成当前价格。
# ========================================================

# 导入模块
import numpy as np
import pandas as pd
import CAL.PyCAL as cal

# 时间参数设定
calender=cal.Calendar('China.SSE') #获取上交所的交易日日历
today=cal.Date.todaysDate() #今日日期
span = cal.Period('5Y') #时间间隔
beginDate=calender.adjustDate(today-span,cal.BizDayConvention.Preceding) 
span = cal.Period('1D') #时间间隔
endDate=calender.adjustDate(today-span,cal.BizDayConvention.Preceding) #根据交易所日历求前一个交易日
month=today.month() #月份
year=today.year() #年份

def fun_avgPE(ticker,beginDate,endDate):
    """
    功能：求某只股票的平均市盈率
    参数：
        ticker(str):股票代码
        beginDate（str):开始日期
        endDate（str）：结束日期
    返回值：
        PE平均值（float）
    """
    df=DataAPI.MktStockFactorsDateRangeGet(ticker=ticker,beginDate=beginDate,endDate=endDate,\
                                       field=u"ticker,tradeDate,PE",pandas="1")
    return df['PE'].mean()

def fun_growRate(ticker,nYear=3):
    """
    功能：统计盈利增长率的平均值
    参数：
        ticker(str):股票代码
        nYear(int):时间跨度，如3表示取3年的增长率
    返回值：avgGrowRate(float)：平均增长率
    备注：由于报表调整导致存在同一年有不同净利润的情况，用groupby().first()整理
    """
    # 为求n年增长率需要获取n+1年的数据
    df=DataAPI.FdmtISGet(ticker=ticker,reportType=u"A",beginYear=str(year-nYear-1),\
                         endYear=str(year-1),field=u"endDate,NIncome",pandas="1")
    df=df.dropna() #去空
    df=df.groupby('endDate').first().reset_index() #同一年只取最新的数据
    avgGrowRate=0
    for i in range(nYear):
        endDate1=str(year-i-1)+'-12-31' #财务截止日期
        endDate2=str(year-i-2)+'-12-31' #财务截止日期
        NIncome1=df[df.endDate==endDate1]['NIncome'].iloc[0]
        NIncome2=df[df.endDate==endDate2]['NIncome'].iloc[0]
        avgGrowRate=avgGrowRate+(NIncome1-NIncome2)/NIncome2 #求增长率并累计
    avgGrowRate=avgGrowRate/nYear #求平均增长率
    return min(avgGrowRate,0.2) #增长率上限20%

def fun_EPS(ticker,year):
    """
    功能：获取当前EPS
    参数：
        ticker(str):股票代码
        year(int):最近一个报告年度
    返回值：EPS(float):最近一个报告年度的EPS
    """
    df=DataAPI.FdmtISGet(ticker=ticker,reportType=u"A",beginYear=str(year-1),\
                         endYear=str(year-1),field=u"ticker,basicEPS",pandas="1")
    return df['basicEPS'][0]

def fun_dividRate(ticker,n=5):
    """
    功能：求红利分配率（平均值）
    参数：
        ticker(str):股票代码
        n(int):年数
    返回值：dividRate(float):红利分配率
    """
    df=DataAPI.FdmtISGet(ticker=ticker,reportType=u"A",beginYear=str(year-n),\
                     endYear=str(year-1),field=u"endDate,basicEPS",pandas="1")
    df=df.dropna().drop_duplicates()
    df=df.groupby('endDate').first().reset_index()

    beginDate=str(year-5)+"-01-01"
    df1=DataAPI.EquDivGet(ticker=u"000001",beginDate=beginDate,endDate=str(today),\
                         field=u"endDate,perCashDiv",pandas="1")
    df1=df1.drop_duplicates().fillna(0)
    
    df2=pd.merge(df,df1,how='inner',on='endDate')
    eps=df2['basicEPS'].sum()
    cashDivid=df2['perCashDiv'].sum()
    return cashDivid/eps

# 以下数据手工输入设定值
ticker=u'000001' #股票代码
expEarnRate=0.15 #期望回报率
nYear=10 # 时间跨度

# 以下数据自动获取
EPS=fun_EPS(ticker,year) #最近一年的每股收益
avgPE=fun_avgPE(ticker,beginDate,endDate) #平均市盈率
growRate=fun_growRate(ticker) #盈利增长率，来源于selection2017.csv
dividRate=fun_dividRate(ticker) #红利分配率

# 以下数据为估值计算的中间值
futureEPS=EPS*(1+growRate)**nYear #最后一年的EPS
totalDivid=0
for i in range(nYear):
    totalDivid=totalDivid+EPS*(1+growRate)**(i+1)*dividRate #累计红利
futurePrice=futureEPS*avgPE+totalDivid #最后一年的预期股价
currPrice=futurePrice/((1+expEarnRate)**nYear) #实现期望回报率的现价（估值）

# 显示计算过程和结果
print(u'股票代码：'+ticker)
print(u'期望回报率：'+str(expEarnRate*100)+"%")
print(u'投资年数：'+str(nYear))
print(' ')
print(u'当前EPS:'+str(EPS))
print(u'平均市盈率：'+str(avgPE))
print(u'盈利增长率(最大20%）：'+str(growRate*100)+'%')
print(u'红利分配率：'+str(dividRate*100)+'%')
print(' ')
print(u"最后一年的EPS:"+str(futureEPS))
print(u'累计红利分配：'+str(totalDivid))
print(u'最后一年的预期股价：'+ str(futurePrice))
print(' ')
print(u'实现期望回报率的当前价格（估值）：'+str(currPrice))
```
