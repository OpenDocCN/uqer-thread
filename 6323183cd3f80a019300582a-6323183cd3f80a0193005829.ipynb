		             筹码分布的计算例子				
		成交均价	换手率（turnoverRate)	没有被重新交易的比率			调整还手率（chouma）
周二	2022-09-06	10	0.1			(1-20%)×（1-30%）×（1-40%）=0.336	0.336
周三	2022-09-07	11	0.2			20%×（1-30%）×（1-40%）=0.084		0.084
周四	2022-09-08	12	0.3			30%×（1-40%）=0.18			0.18
周五	2022-09-09	13	0.4			0.4					0.4
			总换手率：	1	总换手率：	1

换手率的路径依赖调整：以上表为例，周四交易的30%的股份在周五还没有被重新交易的股份为30%×（1-40%）=18%；周三交易的20%的股份在周五还没有被重新交易的股份为20%×（1-40%）

×（1-30%）=8.4%；周二交易的10%股票还没有被重新交易的股份为(1-20%)×（1-40%）×（1-30%）。
换手率归一化，可根据调整换手率的和对每日的调整换手率进行归一化。

```python
import pandas as pd
df = pd.DataFrame(columns=["tradeDate","mean","turnoverRate","close"])
df.loc[0]=["2022-09-06",10,0.1,10.5]
df.loc[1]=["2022-09-07",11,0.2,10.50]
df.loc[2]=["2022-09-08",12,0.3,13.50]
df.loc[3]=["2022-09-09",13,0.4,12.50]
df.index=df["tradeDate"]
print "全还手，总的换手率应该为1:",df["turnoverRate"].sum()
print "0df:\n",df

def get_chouma(df0):
    '''输入：
        dataframe 
        columns=["tradeDate","mean","turnoverRate","close"
        索引为tradeDate
        全还手，总的换手率应该为1
    '''
    df=df0.copy()
    #必须置为df['t'][0] = 1,便于计算调整的换手率即筹码
    df['turnoverRate'][0] = 1
    print "1df:\n",df
    df_turnoverRate = df['turnoverRate']
    print "df_turnoverRate:\n",df_turnoverRate
    df_atr=pd.Series(index=df.index)
    for i in range(len(df_atr)):
        df_atr.iloc[i] = df_turnoverRate.iloc[i] * (1 - df_turnoverRate.iloc[i+1:]).prod()
    #最后一天筹码即为最后1天换手率
    df_atr.iloc[-1]=df["turnoverRate"].iloc[-1]
    print "筹码分布df_atr:"#,df_atr
    return df_atr
df_atr=get_chouma(df)
print df_atr
```
