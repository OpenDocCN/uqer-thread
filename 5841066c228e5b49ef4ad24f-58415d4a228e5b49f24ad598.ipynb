# 轮盘游戏

#### 今天一位同学用代码测试了我在之前一篇帖子中提到一个轮盘游戏，让我觉得这个问题还是有一些可供思考的地方。
具体可以看：[投资组合长期超额收益的根源的最近思考](https://uqer.datayes.com/community/share/57e8bf08228e5b47ec0d8f24)

> 轮盘游戏简介：
> 假设有一个轮盘，轮盘上分成3个区域，每个区域面积不同，，且对应不同奖励。你可以在任何区域下注，下注金额是1块。如果最后指针指向不同区域则无法得到奖励且1块钱无法收回。
第一个区域面积是1/2, 获得的奖励是3块，第二个区域面积是1/3,获得的奖励是2块，第三个区域面积是1/6,获得的奖励是6块。

> 如果是你来玩这个游戏，初始有100块钱，怎么玩才能在N期之后的期望最大收益？

>这时候我们心理就开始算了：第一个的期望收益是1.5 -1 = 0.5，第二个的期望收益是0.667 - 1 = - 0.33，第三个的期望收益是 1 - 1 = 0。那还不简单，就专注于投资第一个区域呗！

> 这时候，有的同学可能还会考虑，用凯利策略来为每次投资第一个区域的资金进行最优分配。毕竟凯利策略也是多期投资下的嘛。但不管怎么样，你都会想去只投资第一个对不对？

>来看看对数效用模型是怎么做的。该模型是把投资者的效用函数假设为对数形式，在此假设下，目标是最大化第N期的期望效用。

>这个模型的结论惊人：你只需要最大化第一个时期的效用函数，并保持这个投资比例，就能最大化第N期的期望效用。而且，甚至最后你都不需要效用函数是对数形式的假设！

>也就是说，你只需要像对待单期投资一样，对待多期投资。

>好吧，是不是感觉好像说了和没说一样，并没有什么特别的感觉？

>那么，我们再来看一下前面提到的那个游戏。在这个理论下，给出的策略很违反我们的直觉：始终把1/2的资金投资在面积为1/2的区域，把1/3的资金投资在面积为1/3的区域，1/6的资金投资在面积为1/6的区域。即使是在预期净收益为负数的第二个区域，也进行了投资。

>不相信的同学，可以用excel或是其他软件简单蒙特卡洛模拟一下50期、100期、200期之后的结果，对比你的策略和这个对数策略，相信你会很吃惊。

>之所以这样投资能长期赚钱的缘由，不完全是因为投资风险分散化，而是由于书中所称的”泵吸"效应，通俗来说就是高抛低吸。哈哈是不是像某强国的庄家所为。。。。



针对那位同学的疑问，我想先说明一下：
1. 在这个策略中，是按照比例投资，100块本金，不是每次只投资6块，这样投资效率也太低了，而且肯定不是最优策略。而是尽量提高效率，多投资。对数最优策略，是100块的1/2,1/3,1/6分别投资在3个轮盘，而与之比较的策略，是1/2的本金只投资区域一的策略，和1/4的本金只投资区域一的策略（这个策略，书上证明了在只投资区域一的极限意义下是最优的）
2. 我自己做蒙特卡洛模拟，做了1000次，然后取均值，发现对数策略远远落后于对手策略。当时我也想了很久不明白。后来我想清楚了，应该是均值收到极端值的影响。对手策略，“偶尔"会产生非常大非常大的极端值，所以一取平均值，会远远超过对数最优策略。
3. 所以，我想了一个方法证明对数最优策略。就是看对数最优策略的胜率winRatio。只要一次模拟结束后，对数最优策略的结果好于对手策略，则winRatio+1,最后，再将winRatio除以总的模拟次数，算出胜率。

下面的程序中，onlyOne是1/2的本金只投资区域一的策略，spreadThree是分散策略。

```python
import random
import numpy as np
onlyOne = 100     #起始金额
spreadThree = 100     # 起始金额
winRatio = 0            #策略二相对策略一胜率
total1 = []
total2 = []
for i in range(1000):              #蒙特卡洛模拟：1000次
    for j in range(50):           #时间范围：50个时期
        rand = random.randint(1,6)
        one1 = onlyOne /2
        onlyOne = onlyOne - onlyOne/2      # 用拥有金额的一半，只投区域一
        
        three1 = spreadThree / 2
        three2 = spreadThree / 3
        three3 = spreadThree / 6
        spreadThree = spreadThree - spreadThree / 6 - spreadThree / 3 - spreadThree / 2       # 分别按比例投
        
        if rand == 1 or rand == 2 or rand == 3:
            onlyOne = onlyOne + one1 * 3
            spreadThree = spreadThree + three1 * 3
        if rand == 4 or rand == 5:
            spreadThree = spreadThree + three2 * 2
        if rand == 6:
            spreadThree = spreadThree + three3 * 6
    
    if spreadThree > onlyOne:
        winRatio = winRatio + 1
        
    
    total1.append(onlyOne)
    total2.append(spreadThree)
    onlyOne = 100      # 从新开始
    spreadThree = 100   #从新开始

money1 = np.array(total1)
money2 = np.array(total2)
print("Only invest in one area strategy, mean return is %.2f  "%(money1.mean()))      # 策略一的期望值
print("invest in three areas strategy, mean return is %.2f  "%(money2.mean()))         #策略二的期望值
print("win ratio is %.2f%% " %(winRatio/1000.0*100))       #策略二相对于策略一的胜率
```

尽管对数最优策略模拟的期望均值远远低于对手策略，但winRatio为88.70%。也就是说，88.7%的模拟结果中，对数最优策略胜于对手的1/2投资策略。
