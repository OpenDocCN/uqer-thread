当年报的每股收益、扣除非经常性损益后的每股收益以及每股净资产三项指标同时为正值时，\*ST股票可以向上交所提出摘帽请求。故\*ST股票年报预计盈利，扭亏为盈，则有大概率摘帽。然而通常在公告公布前，股票已经会有所反应，故我们选取年报预计盈利公告发布前涨幅较小的股票，并设置10%的止盈位。

```python
import time
start = time.clock()
All_info_df = pd.DataFrame({})
counter_before = 0
total = 0
counter_after = 0
for secID in basics['secID']:
    forecast = DataAPI.FdmtEfGet(ticker=u"",secID=secID,reportType=u"",endDate=u"",beginDate=u"",publishDateEnd=u"",publishDateBegin=u"",forecastType="21",field=u"",pandas="1")
    #print forecast['publishDate']
    
    if not forecast.empty:
        for dt in forecast['publishDate']:
            beginDate = datetime.datetime.strptime(dt,'%Y-%m-%d %H:%M:%S') 
            endDate = datetime.datetime.strptime(dt,'%Y-%m-%d %H:%M:%S') + datetime.timedelta(30)
            bd = beginDate.strftime('%Y%m%d') 
            ed = endDate.strftime('%Y%m%d')
            raw = DataAPI.MktEqudGet(tradeDate=u"",secID=secID,ticker=u"",beginDate=bd,endDate=ed,isOpen="",field=u"",pandas="1")
            
            hs300=DataAPI.MktIdxdGet(tradeDate=u"",indexID=u"",ticker=u"000300",beginDate=bd,endDate=ed,field=u"",pandas="1")
            
            
            try:
                hsatr=(hs300['closeIndex'][-1:]-hs300['closeIndex'][0])/hs300['closeIndex'][0]
                rawatr=(raw['closePrice'][-1:]-raw['closePrice'][0])/raw['closePrice'][0]
            
                field1 = ['secID','publishDate','endDate','secIDclosePricebd','secIDclosePriceed','secIDatr','hs300bd','hs300ed','hsatr','negMarketValue']
                df=DataFrame([secID,bd,ed,raw['closePrice'][0],raw['closePrice'][-1:].values,rawatr.values,hs300['closeIndex'][0],hs300['closeIndex'][-1:].values,hsatr.values,raw['negMarketValue'][0]],index=field1)
            
                #print raw['negMarketValue'][0]
                #print secID,bd,ed,raw['closePrice'][0],raw['closePrice'][-1:].values,rawatr.values,hs300['closeIndex'][0],hs300['closeIndex'][-1:].values,hsatr.values,raw['negMarketValue'][0]
                All_info_df=pd.concat([All_info_df,df.T],ignore_index=True)
            except:
                print secID+'错误'
#print counter_before,counter_after,total
end = time.clock()
end - start
```
