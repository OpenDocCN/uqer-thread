一些时间函数，每个人都有自己处理习惯，我比较喜欢拿出一个时间列表，然后在通过列表索引定位。

```python
def get_monthend_trading_day_list(begin_day="2007-01-01", end_day="2017-12-31"):
    """获取每月的最后一个交易日"""
    cal_dates = DataAPI.TradeCalGet(exchangeCD=u"XSHG", beginDate=begin_day, endDate=end_day, field="calendarDate,isMonthEnd")
    trading_days = cal_dates[cal_dates['isMonthEnd']==1]["calendarDate"].tolist()
    return trading_days

def get_monthend(date_str='2016-01-04'):
    """获取该日的月底的最后一个交易日"""
    begin_date = datetime.strptime(date_str, '%Y-%m-%d')
    end_date = begin_date + 100 * timedelta(days=1)
    begin_day_str = begin_date.strftime('%Y-%m-%d')
    end_day_str = end_date.strftime('%Y-%m-%d')
    return get_monthend_trading_day_list(begin_day_str, end_day_str)[0]

#获得必要的连续两个月底交易日，文本
def month_end_to_factor(input_first_date,date_list):
    get_next_end=cal.advanceDate(input_first_date,'-1B',BizDayConvention.Preceding).strftime('%Y-%m-%d')
    get_next_next_end=date_list[date_list.index(get_next_end)-1]
    return get_next_end,get_next_next_end

#获得每月最后一个交易日文本型列表
month_num=['01-15','02-15','03-15','04-15','05-15','06-15','07-15','08-15','09-15','10-15','11-15','12-15']
year=list(range(2009,2018))
date_list=[]
for y in year:
    for m in month_num:
        date_list.append(get_monthend(str(y)+'-'+m))        
```

由于是关注每个具体行业，所以免去了行业中性化，因子去极值和标准化完全可以直接写。这里数据是建立在Pandas的DataFrame和Series基础上的。

```python
#3倍标准差法去极值
def winsize(x,input_series):
    base_compare_up=input_series.mean()+3*input_series.std()
    base_compare_down=input_series.mean()-3*input_series.std()
    if x > base_compare_up:
        x=base_compare_up
    elif x<base_compare_down:
        x=base_compare_down
    else:
        x=x
    return x
#标准化
def standarize(y,input_series_std):
    i_mean=input_series_std.mean()
    i_std=input_series_std.std()
    y_s=(y-i_mean)/i_std
    return y_s
```

1、buylist函数里，ticker_input 输入申万一级行业
2、factor_input_pool 选择所要的因子，因子最好涉及估值、营收质量、成长、动量、技术类等几个大类，可以在大类中精挑细选，当然最好结合单子测试
3、这里每个月回归一次，对多因子模型动态调整
4、结果是字典形式，键是secID,值是因子总得分
5、在仓位重采用等权重，可以根据因子得分重新进行设置

```python
def buylist(input_date,date_list):
    ticker_input=u"801080"
    factor_input_pool=u"secID,ETOP,ROE,OperatingRevenueGrowRate,LCAP,BLEV,REVS20,VOL20,RSI"#,PCF,GrossIncomeRatio"
#get need time
    next_end=month_end_to_factor(input_date,date_list)[0]
    next_end_pre=month_end_to_factor(input_date,date_list)[1]
#get stock pool
    med_stock=DataAPI.IdxConsGet(secID=u"",ticker=ticker_input,intoDate=next_end_pre,\
                                 field=u"consID,consTickerSymbol,consShortName",pandas="1").set_index('consID')
    med_stock_pool_secID=med_stock.index.tolist()

    #剔除上市不满一年的   注意时间设置
    basic_info=DataAPI.EquGet(equTypeCD=u"A",secID=med_stock_pool_secID,field=u"secID,ticker,secShortName,listDate",pandas="1").set_index('secID')
    basic_info=basic_info[basic_info.listDate<(str(int(next_end_pre[:4])-1)+next_end_pre[4:])]
    med_stock_pool=basic_info.index.tolist()
#get all factor for regress
    factor_get=DataAPI.MktStockFactorsOneDayGet(tradeDate=next_end_pre,secID=med_stock_pool,\
                                                field=factor_input_pool,pandas="1").set_index('secID')
#win and standar
    for cl in factor_get.columns:
        factor_get['factor_win_'+cl]=factor_get[cl].apply(lambda x:winsize(x,factor_get[cl]))
        del factor_get[cl]
    factor_get_win=factor_get
    for clst in factor_get_win.columns:
        factor_get_win[clst+'_std']=factor_get_win[clst].apply(lambda x:standarize(x,factor_get_win[clst]))
        del factor_get_win[clst]
    factor_get_win_std=factor_get_win
#get all stocks' retun
    stock_price=DataAPI.MktEqumAdjGet(secID=med_stock_pool,monthEndDate=next_end,isOpen=u"1",field=u"secID,chgPct",pandas="1").set_index('secID')
    factor_get_win_std['m_return']=stock_price.chgPct
    factor_get_win_std=factor_get_win_std.dropna()
#regress for getting coe
    X=factor_get_win_std[factor_get_win_std.columns[:-1]]
    Y=factor_get_win_std[factor_get_win_std.columns[-1]]
    est=sm.RLM(Y,X).fit()
    coe=est.params.tolist()
    
#get new factor for ranking by coe
    factor_get_new=DataAPI.MktStockFactorsOneDayGet(tradeDate=next_end,secID=med_stock_pool,\
                                                field=factor_input_pool,pandas="1").set_index('secID')
#win and standar
    for cl in factor_get_new.columns:
        factor_get_new['factor_win_'+cl]=factor_get_new[cl].apply(lambda x:winsize(x,factor_get_new[cl]))
        del factor_get_new[cl]
    factor_get_win_new=factor_get_new
    for clst in factor_get_win_new.columns:
        factor_get_win_new[clst+'_std']=factor_get_win_new[clst].apply(lambda x:standarize(x,factor_get_win_new[clst]))
        del factor_get_win_new[clst]
    factor_get_win_std_new=factor_get_win_new
    
#score plus
    all_name=list(factor_get_win_std_new.columns)
    for col in all_name:
        factor_get_win_std_new['coe_'+col]=factor_get_win_std_new[col]*coe[all_name.index(col)]
        del factor_get_win_std_new[col]
    
    rank_factor_series=factor_get_win_std_new.T.sum()
    rank_factor_by_big=rank_factor_series.sort_values(ascending=False)
    
#change to dict
    buylist_dict=rank_factor_by_big[:10].to_dict()
    
    return buylist_dict
```
