本文参考申万研报[《申万大师系列价值投资篇之七：彼得.林奇基层调查选股法》](http://vdisk.weibo.com/s/tZKrQTy0H2oP?from=page_100505_profile&wvr=6)，对大师的思想进行一些量化。具体量化了以下5条：
1. 公司的资产负债率小于等于25%
2. 公司每股净现金大于0
3. 当前股价与每股自由现金流量比小于10
4. 公司的存货增长率小于其营收增长率
5. (长期盈余增长率+股息率)/市盈率大于等于2

```python
#过滤条件：公司的资产负债率小于等于25%
def debtToasset(account):
    cal = Calendar('China.SSE')
    yestoday = cal.advanceDate(account.current_date,'-1B',BizDayConvention.Unadjusted)
    eq_DebtToAsset = pd.DataFrame()
    for s in xrange(300,len(account.universe),300): #由于API的读取限制，分批运行API。
        tmp = DataAPI.MktStockFactorsOneDayGet(secID=account.universe[s-300:s],tradeDate=yestoday.strftime("%Y%m%d"),field=['secID','DebtsAssetRatio'],pandas="1")
        eq_DebtToAsset = pd.concat([eq_DebtToAsset,tmp],axis = 0)
    s = (len(account.universe) / 300)*300
    tmp = DataAPI.MktStockFactorsOneDayGet(secID=account.universe[s:],tradeDate=yestoday.strftime("%Y%m%d"),field=['secID','DebtsAssetRatio'],pandas="1")
    eq_DebtToAsset = pd.concat([eq_DebtToAsset,tmp],axis = 0)   

    eq_filter1 = eq_DebtToAsset[eq_DebtToAsset['DebtsAssetRatio'] < 0.25]['secID'].tolist()
    return eq_filter1
```

```python
#过滤条件：公司每股净现金大于0与当前股价与每股自由现金流量比小于10
def ncfOperateA(account,filter_list):
    today = account.current_date
    cal = Calendar('China.SSE')
    lastDay = cal.advanceDate(today,'-1B',BizDayConvention.Unadjusted)
    today_str = lastDay.strftime("%Y%m%d")
    start_date = cal.advanceDate(lastDay,'-500B',BizDayConvention.Unadjusted)
    start_date_str = start_date.strftime("%Y%m%d")

    equity = pd.DataFrame()
    for rpt_type in ['Q1','S1','A']:
        try:
            tmp = DataAPI.FdmtCFGet(secID=filter_list,field=['secID', 'endDate', 'publishDate', 'NCFOperateA'], beginDate=start_date_str, publishDateEnd=today_str,  reportType=rpt_type)
        except:
            tmp = DataFrame()
        equity = pd.concat([equity, tmp], axis=0)
        
    equity = equity.sort(columns=['secID','endDate','NCFOperateA'], ascending=[True, False, False])
    equity = equity.drop_duplicates(cols=['secID','endDate'])
    equity = equity.dropna()
    
    signalNCFO = {}
    for stock in filter_list:
        try:
            num = len(equity[equity['secID']==stock]['endDate'])
            if (num>=4):
                endDate = equity[equity['secID']==stock]['endDate'].iloc[0]
                month= int(endDate[6])
                if (month==2) :
                    NCFO_TTM = equity[equity['secID']==stock]['NCFOperateA'].iloc[0]
                elif  (month==6):
                    NCFO_TTM = equity[equity['secID']==stock]['NCFOperateA'].iloc[0]-equity[equity['secID']==stock]['NCFOperateA'].iloc[3]+equity[equity['secID']==stock]['NCFOperateA'].iloc[2] 
                elif (month==3):
                    NCFO_TTM = equity[equity['secID']==stock]['NCFOperateA'].iloc[0]-equity[equity['secID']==stock]['NCFOperateA'].iloc[3]+equity[equity['secID']==stock]['NCFOperateA'].iloc[1]
                else:
                    NCFO_TTM = 'NA'
                if len(signalNCFO) == 0:
                    signalNCFO['secID'] = [stock]
                    signalNCFO['signalNCFO'] = [NCFO_TTM]
                else:
                    signalNCFO['secID'].append(stock)
                    signalNCFO['signalNCFO'].append(NCFO_TTM)    
        except:
            pass
 
    signalNCFO = DataFrame(signalNCFO)
    filterList1 = signalNCFO['secID'].tolist()
    signalNCFO =  signalNCFO[signalNCFO['signalNCFO'] > 0] 
    
    filter1_NCFO = {}
    for stock in filterList1:
        try:
            tmp = DataAPI.MktEqudGet(secID=stock,tradeDate=today_str,field='marketValue',pandas="1")
            price_NCFO = tmp['marketValue'].values[0]/signalNCFO[signalNCFO['secID'] == stock]['signalNCFO'].values[0]

            if len(filter1_NCFO) == 0:
                filter1_NCFO['secID'] = [stock]
                filter1_NCFO['price_NCFO'] = [price_NCFO]
            else:
                filter1_NCFO['secID'].append(stock)
                filter1_NCFO['price_NCFO'].append(price_NCFO)    
        except:
            pass

    filter1_NCFO = DataFrame(filter1_NCFO)
    
    filter1_NCFO = filter1_NCFO[filter1_NCFO['price_NCFO'] < 10]
    filterList2 = filter1_NCFO['secID'].tolist()
    return filterList2
```

```python
#过滤条件：公司的存货增长长率小于其营收增长率
def inv_oper(account,filter_list):
    today = account.current_date
    cal = Calendar('China.SSE')
    lastDay = cal.advanceDate(today,'-1B',BizDayConvention.Unadjusted)
    today_str = lastDay.strftime("%Y%m%d")
    start_date = cal.advanceDate(lastDay,'-500B',BizDayConvention.Unadjusted)
    start_date_str = start_date.strftime("%Y%m%d")


    equity = pd.DataFrame()
    for rpt_type in ['Q1','S1','A']:
        try:
            tmp = DataAPI.FdmtBSGet(secID=filter_list,field=['secID', 'endDate', 'publishDate', 'inventories'], beginDate=start_date_str, publishDateEnd=today_str,  reportType=rpt_type)
        except:
            tmp = DataFrame()
        equity = pd.concat([equity, tmp], axis=0)

    equity = equity.sort(columns=['secID','endDate','inventories'], ascending=[True, False, False])
    equity = equity.drop_duplicates(cols=['secID','endDate'])
    equity = equity.dropna()
    
    inv_growth = {}
    
    for stk in filter_list:
        try:
            inv_growth[stk] = equity[equity['secID'] == stk]['inventories'].values[0] / equity[equity['secID'] == stk]['inventories'].values[3] - 1
        except:
            pass
    
    eq_revgrow = DataAPI.MktStockFactorsOneDayGet(tradeDate=today_str,secID=filter_list,field=['secID','OperatingRevenueGrowRate'],pandas="1")
    
    filtered_list = []
    
    for stk in filter_list:
        try:
            if eq_revgrow[eq_revgrow['secID'] == stk]['OperatingRevenueGrowRate'].values > iv_growth[stk]:
                filtered_list.append(stk)
        except:
            filtered_list.append(stk)

    return filtered_list
```

可以看出回测结果是完全一样的，表明回测区间内满足前3个条件的股票也满足第四个条件。

```python
#过滤条件：(5 年平均净利润同比增长率 + 股息率)/市盈率 > 2 PS:之前有计算过股息率，基本都很小，对比市盈率，基本可以忽略。
def nIncome_avgrowth(account,filter_list):
    today = account.current_date
    cal = Calendar('China.SSE')
    lastDay = cal.advanceDate(today,'-1B',BizDayConvention.Unadjusted)
    today_str = lastDay.strftime("%Y%m%d")
    start_date = cal.advanceDate(lastDay,'-2000B',BizDayConvention.Unadjusted)
    start_date_str = start_date.strftime("%Y%m%d")


    equity = pd.DataFrame()
    for rpt_type in ['Q1','S1','A']:
        try:
            tmp = DataAPI.FdmtISGet(secID=filter_list,field=['secID', 'endDate', 'publishDate', 'NIncome'], beginDate=start_date_str, publishDateEnd=today_str,  reportType=rpt_type)
        except:
            tmp = DataFrame()
        equity = pd.concat([equity, tmp], axis=0)

    equity = equity.sort(columns=['secID','endDate','NIncome'], ascending=[True, False, False])
    equity = equity.drop_duplicates(cols=['secID','endDate'])
    equity = equity.dropna()
    netIncome_growth = {}
    
    avg_growth = 0
    #计算5 年平均净利润同比增长率
    for stk in filter_list:
        try:
            for i in xrange(0,15,3):
                avg_growth  = avg_growth + equity[equity['secID'] == stk]['NIncome'].values[i] / equity[equity['secID'] == stk]['NIncome'].values[i+3] - 1
                
            netIncome_growth[stk] = avg_growth / 5
        except:
            pass 
     
    eq_pe = DataAPI.MktStockFactorsOneDayGet(tradeDate=today_str,secID=filter_list,field=['secID','PE'],pandas="1") 
    
    filtered_list = []
    
    for stk in filter_list:
        try:
            if netIncome_growth[stk] / eq_pe[stk] > 2:
                filtered_list.append(stk)
        except:
            filtered_list.append(stk)
    return filtered_list
```
