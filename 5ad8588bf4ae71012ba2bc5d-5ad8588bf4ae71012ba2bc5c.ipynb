```python
# 复制的坑
# case 1
a = [1, 2, 3]
b = a
print(a, b, id(a), id(b))
b[0] = 11
print(a, b, id(a), id(b))
```

```python
# case 2
import copy 
a = [1, 2, 3]
b = copy.copy(a)
print(a, b, id(a), id(b))
b[0] = 11
print(a, b, id(a), id(b))
```

```python
# case 3
a = (1, 2, 3, [4, 5])
b = copy.copy(a)
print(a, b, id(a), id(b))
b[3][0] = 11
print(a, b, id(a), id(b))
```

```python
# case 4
a = (1, 2, 3, [4, 5])
b = copy.deepcopy(a)
print(a, b, id(a), id(b))
b[3][0] = 11
print(a, b, id(a), id(b))
```

小结:   python的变量定义  浅拷贝  深拷贝

### 3.3 填充空缺值

```python
industry_mean = df_merge.groupby('industryName1').agg({'CurrentAssetsRatio':'mean', 'CTP5':'mean', 'NetAssetGrowRate':'mean'})
industry_mean
```

```python
# 简单方法，按照逻辑一步步走
factors = ['NetAssetGrowRate', 'CurrentAssetsRatio', 'CTP5']
for industry in industry_mean.index:
    # for factor in factors:
    #     df_merge.loc[df_merge['industryName1'] == industry, factor] = df_merge.loc[df_merge['industryName1'] == industry, factor].fillna(industry_mean.loc[industry][factor])
    df_merge.loc[df_merge['industryName1'] == industry, factors] = df_merge.loc[df_merge['industryName1'] == industry, factors].fillna(industry_mean.loc[industry][factors])
```

```python
# 更简便方法
tmp_df.loc[:, factors] = tmp_df.loc[:, factors].fillna(tmp_df.groupby('industryName1')[factors].transform("mean"))
```
