# 各行业费用占比：总体与医疗生物行业内费率结构

```python
import pandas as pd
sec=set_universe('A')
df=DataAPI.EquIndustryGet(secID=sec,industryVersionCD=u"010303",industry=u"",industryID=u"",industryID1=u"",industryID2=u"",industryID3=u"",intoDate=u"20181231",field="endDate,ticker,industryName1",pandas="1")
sec1=df[df['industryName1']=='医药生物'].set_index('ticker').index.tolist()
df1=DataAPI.FdmtISGet(ticker=sec1,reportType=u"A",endDate=u"",beginDate=u"",publishDateEnd=u"",publishDateBegin=u"",endDateRep="",beginDateRep="",beginYear="2018",endYear="2018",fiscalPeriod="",field=u"endDate,ticker,tRevenue,COGS,sellExp,adminExp,finanExp",pandas="1")
df2=DataAPI.EquIndustryGet(ticker=sec1,industryVersionCD=u"010303",industry=u"",industryID=u"",industryID1=u"",industryID2=u"",industryID3=u"",intoDate=u"20181231",field="endDate,ticker,industryName3",pandas="1")
df=df1.merge(df2,left_on=['ticker'], right_on=['ticker'], how='inner')
df['xscbl']=df['COGS']/df['tRevenue']
df['ssfyl']=df['sellExp']/df['tRevenue']
df['glfyl']=df['adminExp']/df['tRevenue']
df['cwfyl']=df['finanExp']/df['tRevenue']
df[['ticker','xscbl','ssfyl','glfyl','cwfyl','industryName3']]
fyjg=pd.DataFrame(index=df['industryName3'].drop_duplicates(),columns=[['xscbl','ssfyl','glfyl','cwfyl']])
for x in df['industryName3'].drop_duplicates():
    fyjg.loc[x]=df[df['industryName3']==x][['xscbl','ssfyl','glfyl','cwfyl']].mean()
print fyjg
for y in ['xscbl','ssfyl','glfyl','cwfyl']:
    print fyjg[y].argmax()
```
