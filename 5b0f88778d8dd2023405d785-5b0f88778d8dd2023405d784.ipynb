```python
#2、计算贵州茅台与恒瑞医药2018年2月所有交易日的TTM营业总收入和TTM管理费用
#思路：
#1、获取财报数据
#2、只取T日及之前发布的的财报,选取发布的累计值信息
#3、排序、去重，选取最近一个财报期的财报，记录其财报期为A
#4、选取A，A之前一年的年报记为B，去年与A同期的财报记为C
#5、获取A、B、C中的所需科目数据
#6、循环遍历2月份的交易日，计算出每一交易日的数据，营业总收入和管理费用

#代码实现：
#1、构建计算科目在不超过T日的TTM的子函数
#2、构建获取交易日期的子函数
#3、构建计算一段时间多只股票的某一TTM
#4、合并不同的TTM
import pandas as pd
from dateutil.relativedelta import relativedelta

def XX_TTM(ticker_symbol,T,XX,read_start_date='2015-01-01'):
    income = DataAPI.FdmtISGet(ticker=ticker_symbol,secID=u"",reportType=u"",endDate=u"",beginDate=u"",publishDateEnd=u"",publishDateBegin=u"",endDateRep="",beginDateRep=read_start_date,beginYear="",endYear="",fiscalPeriod="",field=u"ticker,publishDate,endDate,fiscalPeriod,tRevenue,adminExp",pandas="1")
    income['publishDate'] = pd.to_datetime(income['publishDate'])#时间格式统一，后面方便操作
    income['endDate'] = pd.to_datetime(income['endDate'])
    income = income.loc[income['publishDate'] <= T]#保证不取未来时点的数据
    
    df = income.loc[income['endDate'].dt.month == income['fiscalPeriod'].astype(int)]#取累计值，剔除单季度数据
    df = df.sort_values(by=['endDate', 'publishDate'], ascending=[False, False]).drop_duplicates(subset=['endDate'], keep='first')#分别按'endDate', 'publishDate'降序排列，根据'endDate'去重，保留第一次出现的结果，去重后保留某一财报期最新发布的数据
    #df = df.fillna(method='bfill')#如果遇到缺失值用后面的值来填充
    latest_date = df.iloc[0].loc['endDate']#获取最近的一次财报期
    
    last_year_end_date = pd.to_datetime(str(latest_date.year - 1) + '1231')
    last_same_season_end_date = latest_date-relativedelta(years=1)
    
    df = df.set_index('endDate')#把财报期设置为索引
    income_this_season = df.loc[latest_date.date()].loc['{}'.format(XX)]#获取对应科目在指定时间的数据
    income_last_year = df.loc[last_year_end_date.date()].loc['{}'.format(XX)]
    income_last_same_season = df.loc[last_same_season_end_date.date()].loc['{}'.format(XX)]
    XX_TTM = income_this_season + income_last_year - income_last_same_season#根据获得的科目数据进行计算TTM
    
    return XX_TTM

def trade_cal(T1,T2):
    #获取交易日期
    trade_cal = DataAPI.TradeCalGet(exchangeCD=u"XSHG",beginDate=u"",endDate=u"",field=u"isOpen,calendarDate",pandas="1")
    trade_cal = trade_cal[trade_cal['isOpen']==1]
    trade_cal['calendarDate'] = pd.to_datetime(trade_cal['calendarDate'])
    trade_cal = trade_cal.loc[trade_cal['calendarDate'].between(pd.to_datetime(T1), pd.to_datetime(T2))]
    return trade_cal

    
def cal_all_TTM(ticke_list,XX,T1,T2):
    #获取多支股票的在指定时间段的某一科目TTM
    TTM_df = pd.DataFrame(columns=['ticker', 'tradeDate']+['TTM_{0}'.format(XX)])
    for ticker_symbol in ticke_list:
        for trade_date in trade_cal(T1,T2)['calendarDate']:
            ttm = XX_TTM(ticker_symbol,trade_date,XX,read_start_date='2015-01-01')
            TTM_df.loc[len(TTM_df), :] = [ticker_symbol, trade_date, ttm]#此处的loc[len(acca_df),:]用的非常好
    TTM_df = TTM_df.sort_values(by = ['tradeDate','ticker'])
    return TTM_df


ticke_list = ['600519', '600276']
T1 = '20180201'
T2 = '20180228'
# print cal_all_TTM(ticke_list,'tRevenue')
# print cal_all_TTM(ticke_list,'adminExp')
all_TTM = pd.merge(cal_all_TTM(ticke_list,'tRevenue',T1,T2),cal_all_TTM(ticke_list,'adminExp',T1,T2),left_on=['ticker', 'tradeDate'], right_on=['ticker', 'tradeDate'],how='left')
all_TTM
```
