```python
import datetime
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import cvxopt as opt
from CAL.PyCAL import *
from pandas import DataFrame, Series
from cvxopt import matrix, solvers, blas            #凸优化包cvxopt

startDate = '2015-09-03'                       # 回测起始时间
endDate = '2017-12-01'

fund_list = ["510050.XSHG","518880.XSHG","159920.XSHE","511880.XSHG","511010.XSHG"]
listField = ['secShortName','tradeDate', 'closePrice','openPrice']

def get_fundData(fund_code):
    cal = Calendar('China.SSE') 
    data_start=cal.advanceDate(startDate,'-60B',BizDayConvention.Preceding) #60天回测
    fundData = DataAPI.MktFunddAdjBfGet(secID = fund_code,beginDate = data_start, endDate = endDate, field = listField,pandas = "1")
    fundData=fundData.set_index('tradeDate').loc[:,'openPrice'].pct_change()
    
    return fundData.iloc[1:]



fundData=pd.concat(list(map(get_fundData,fund_list)),axis=1)
fundData.columns= fund_list
fundData=fundData.mask(fundData==np.inf)
fundData.fillna(value=0,inplace=True)

fundData.corr()

```

```python
import seaborn
fundData.index= pd.to_datetime(fundData.index,infer_datetime_format=True)
fundData = fundData+1
fundData.iloc[0,:] = 1
fundData=fundData.cumprod(axis=0)
fundData.plot()
plt.show()
```
