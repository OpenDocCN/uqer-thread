# 一级行业对比研究 #
## 1.指数对比 ##
说明：采用申万一级行业指数与沪深300指数做对比

```python
df_L1=df_industry[['一级名称','一级代码','一级指数','一级ID']].drop_duplicates()
#获取指数
index_L1=list(df_L1['一级指数'])
index_L1=map(str,index_L1)
index_L1.append('000300')#沪深300
#定义起止日期
start_date='20140101'
end_date='20170101'
```

```python
#获取期间指数值
df_price=get_index_price_month(start_date,end_date,index_L1)
#获取相对于开始时间的涨幅exceed便于对比，获取最终收益率及月胜率（相对于基准）rank
exceed,rank=month_stastic(df_price)
```

```python
df_L1['一级指数']=df_L1['一级指数'].astype(str)
```

```python
df_result=pd.merge(df_L1,rank,left_on='一级指数',right_index=True)
```

```python
#说明：return：最终收益率，win_rate:月胜率
df_result.sort_values(by='return',ascending=False)
```

```python
#收益走势对比
exceed.sort_index(ascending=True).plot(figsize=[15,10])
```
