看到之前国研网推送的一篇论文《保险资产最优配置:理论模型、数值模拟及政策含义》很感兴趣，就想根据论文的思路来还原他们的分析过程，顺便在这个过程中更加深入地学习Black-Litterman模型。但在试图还原的过程中，我发现无法获得论文中用到的一些数据，而且论文里的一些处理方式不是很直观，我又查阅了一些关于Black-Litterman的论文，感觉其中思路最为清晰的来自于Wing CHEUNG的The Black-Litterman Model Explained （http://www.andreisimonov.com/NES/Litterman_Nomura.pdf）。
简单的讲，BL模型提供了一种将投资者对于资产回报的观点与市场均衡回报水平结合起来的一种方法，它的一大特点就是以均衡水平为出发点，通过Bayes方法形成关于资产回报的后验信息，结具体的分析过程可以分成以下几个步骤：
（1）、投资者形成自己的投资观点及其置信度，这些观点可能来自于投资者的经验也可以来自一些数量模型做出的预测，既可以是绝对观点（如魔衣资产的回报水平），也可以是相对观点（某一类资产的回报超过另一类资产的水平）。
（2）、计算资产的市场均衡水平，有两种常用的方法。一是选取一指数代表市场指数，用CAPM模型计算出的回报水平代表市场均衡水平；二是假设当前的资产权重是市场均衡水平，然后根据市场均衡的权重使用Markowitz的Mean-Variance模型倒推出市场均衡水平下的投资回报。
（3）、将计算出的投资者观点和市场均衡水平代入Black-Litterman模型计算出资产回报水平及其方差的后验值。
（4）、根据后验的资产回报水平及其方差值，采用Mean-Variance模型计算出最优的资产配置权重。

因此，本文借鉴《保险资产最优配置:理论模型、数值模拟及政策含义》一文的思路，采用VAR模型构建基于宏观指标的投资者观点模型，由于关于一些资产品种的市场权重数据无法获得，就采用基于CAPM模型来计算结果来作为市场均衡水平。同时为了看一下效果，按月滚动应用Black-Litterman模型计算最优权重和对应的资产组合的回报并做简单的统计分析,算是一个回测过程。

首先，获取了几类资产的日价格数据，根据数据的可得性，这里主要考虑了股票（上证指数）、债券（中证全债指数）、股票型基金、债券型基金、混合型基金，并取沪深300指数作为参考指数（benchmark）或者说市场指数（market index）。根据每日收盘价采用对数收益率计算日收益率，按月汇总为月收益率。
根据月收益率的图形可以看出，一些资产的月收益率最高会超过20%，最低也可能超过-20%，国内市场的波动可见一斑。

```python
from __future__ import division
import pandas as pd
import numpy as np
from pandas import DataFrame

startDate = u'20030101'
endDate = u'20170531'


stkIdx = '000001'                #①股票：采用上证综合指数(000001)
bondIdx = 'H11001'               #中证全债
stkFundIdx = 'H11021'            #⑤股票型基金：采用中证股票基金指数(H11021)
bondFundIdx = 'H11023'           #⑥债券型基金：采用中证债券基金指数(H11023)
hybridFundIdx = 'H11022'         #⑦混合型基金：采用中证混合基金指数(H11022)
hs300 = '399300'                 #沪深300指数
assetPool = [stkIdx, bondIdx, stkFundIdx, bondFundIdx, hybridFundIdx, hs300]

#industry = DataAPI.IndustryGet(industryVersion=u"SW",industryVersionCD=u"",industryLevel=u"1",isNew=u"1",field=u"industryID,industryName,indexSymbol",pandas="1")
#assetPool = list(industry.indexSymbol.astype('string'))

priceData = DataAPI.MktIdxdGet(indexID=u"",ticker=assetPool,beginDate=startDate,endDate=endDate,field=u"",pandas="1")

indexRets = None
for secId in assetPool:
    retData = priceData[priceData['ticker'] == secId][['tradeDate', 'preCloseIndex', 'closeIndex']]
    retData['logChgPct'] = np.log(retData.closeIndex) - np.log(retData.preCloseIndex)
    retData['tradeDate'] = pd.to_datetime(retData['tradeDate'],format='%Y-%m-%d')
    retData.set_index('tradeDate', inplace = True)
    retData = retData.drop(['preCloseIndex', 'closeIndex'], axis = 1)
    retData.columns = [secId]
    
    if(indexRets is None):
        indexRets = retData
    else:
        indexRets = indexRets.merge(retData, how = 'left', left_index = True, right_index = True)
        
indexRets.columns = ['stkIdx', 'bondIdx', 'stkFundIdx', 'bondFundIdx', 'hybridFundIdx', 'hs300']      
print(indexRets.head(5))
indexMonthlyRets = indexRets.groupby(pd.TimeGrouper(freq='M')).sum()
indexMonthlyRets.plot()
assetMonthlyRets = indexMonthlyRets[range(indexMonthlyRets.shape[1] - 1)]
```

由于VAR模型一般用来对平稳数据进行预测，所以这里对收益率数据进行平稳性检验，ADF测试结果显示，在1%的置信度下，所有资产月收益率非平稳的假设被拒绝，也就是说可以认为这些月收益率数据是平稳的。

```python
from statsmodels.tsa.stattools import adfuller
print('p-value of ADF-test for each asset')
for k in range(assetMonthlyRets.shape[1]):
    adfRes = adfuller(assetMonthlyRets.ix[:,k])
    print(assetMonthlyRets.columns[k] + ': ' + str(round(adfRes[1], 3)))
```

目前用到的python平台上的VAR模型还不支持外生变量，为了简单，本文将所有的变量都视作内生变量，并在这里选择3阶的滞后项。
然后用前100个月的数据对这些VAR模型进行格兰杰因果检验。在检验中我发现，加入“社会融资规模”的两个指标，关于股票指数的方程（第一个方程）能通过格兰杰因果检验，但结果却是后来对债券收益率的预测有着比较大的向下的偏差。所以感觉债券和股票似乎不适合使用一套共同的自变量系统，应该分开进行变量选择和检验。但目前只是先实现一个基本的框架，暂时先简单地把这些收益率放到一个VAR系统里来预测。而且通过对比，投资者观点对结果的影响还是很大的，个人觉得对于这部分也有很大的发挥空间，比如说引进一些更复杂和精致的预测模型。事实上，我最开始的时候误用了“月_消费者景气指数_消费者信心指数”这一指标，模型最后的回测结果很不错，后来才发现这个指标可能会滞后好几个月才发布，所以只好舍弃。

```python
import numpy as np
import pandas
import statsmodels.api as sm
from statsmodels.tsa.api import VAR
mdata.index = pd.to_datetime(mdata.index)

trainData = mdata.iloc[:100]
assets = mdata.columns[:5].tolist()
macroIndicators = mdata.columns[5:]

informativeIndicators = []
for macroIndicator in macroIndicators:
    colSel = assets + [macroIndicator]
    trsamp = trainData[colSel]
    trsampModel = VAR(trsamp).fit(5)
    
    isInformative = False
    for asset in assets:
        #grangerRes = trsampModel.test_causality(asset, trsamp.columns.drop(asset), verbose = False)
        grangerRes = trsampModel.test_causality(asset, macroIndicator, verbose = False)
        if grangerRes.get('pvalue') <= 0.1:
            isInformative = True
    
    if isInformative:
        informativeIndicators.append(macroIndicator)
print('Informative indicators after granger-causality test...')
print(informativeIndicators)

mdata = mdata[assets + informativeIndicators]
```

根据fpe准则选取滞后的阶数为1，从第101个月开始，滚动使用前100个月的数据拟合VAR(1)模型并对下一个月做出预测，记录下所有的预测值及预测的方差。
预测的结果相对来说比较平稳一些，与实际值也有比较大的差距。

```python
plag = 1
predData = None
windowSize = 100

predDate = []
forecastCov = []
forecastCovShape = [mdata.shape[1]]*2
for k in range(windowSize, len(mdata)):
    sample = mdata.iloc[(k-windowSize):k]
    model = VAR(sample).fit(plag)
    pred = model.forecast(sample.values[-plag:], 1)
    forecastCov.append(np.array(pd.DataFrame(model.forecast_cov()[0]).unstack()))
    predRet = pred[0][:5]
    predDate.append(mdata.index[k])
    if predData is None:
        predData = pd.DataFrame([predRet])
    else:
        predData = predData.append([predRet])

predData.set_index(pd.to_datetime(predDate), inplace = True)   
predData.columns = mdata.columns[:5].map(lambda x : x + '_pred')

forecastCov = pd.DataFrame(forecastCov)
forecastCov.set_index(pd.to_datetime(predDate), inplace = True)  

for index in mdata.columns[:5]:
    pd.merge(mdata[str(index)].to_frame(), predData[str(index) + '_pred'].to_frame(),  left_index = True, right_index = True).plot()
```

本文中选取当时的1年期定期存款利率作为无风险利率的参考值并简单地换算称对应的月数据，而不是取一个固定值。这样处理是否合理还优待探索。

```python
startDate = u'20020101'
#M120000007	日_人民币存款基准利率(不定期)_定期存款利率_1年(整存整取)
depInterest = DataAPI.ChinaMacroDataGet(indicID='M120000007',beginDate = startDate, endDate = endDate, field=u"",pandas="1")

depInterest.index = pd.to_datetime(depInterest.publishTime.apply(lambda s: s[:7])) + pd.tseries.offsets.DateOffset(months=1)
depInterest = depInterest[['dataValue']]
depInterest.columns = ['interest']

indexMonthlyRets.index = pd.to_datetime(indexMonthlyRets.index.strftime('%Y-%m'))

tdata = indexMonthlyRets.merge(depInterest[['interest']], how = 'outer', left_index = True, right_index = True)
tdata.interest.fillna(method = 'ffill', inplace = True)
tdata.interest = tdata.interest/12/100
```

好了，开始使用逐月使用Black-Litterman模型了。
首先滚动使用前100个月的数据构计算各个资产的Beta，然后用100个月的市场指数的均值预测各个资产的市场均衡收益率，再结合之前的VAR(3)的预测值，代入Black-Litterman模型计算出资产收益率及其方差的后验值。最后根据资产收益率及其方差的后验值，采用Markowitz的均值方差模型使用最优化方法计算出最优权重。
这里，风险厌恶系数lmbda选取自《保险资产最优配置:理论模型、数值模拟及政策含义》，它和A STEP-BY-STEP GUIDE TO THE BLACK-LITTERMAN MODEL一文中的3.07也比较接近。另一个比较重要的系数是tao，tao越小约偏向市场均衡值，这里取0.05。
由于这里的投资者观点使用的是资产收益率的预测值，而不是相对值，所以，P就是单位阵。而对于描述观点的不确定性的Omega，原本就指观点误差的方差，因此这里就直接取VAR预测模型的方差。

```python
import statsmodels.api as sm
import scipy.optimize as sco

pdata = tdata.dropna().merge(predData, how = 'left', left_index = True, right_index = True)
predStartN = pdata.stkIdx_pred.index.get_loc(predDate[0])

lmbda = 2.848
tao = 0.05

def negPortfolioExpectedUtility(weight, *retCov):
    w = np.mat(weight)
    expectedUtility = w*retCov[0] - lmbda/2*w*retCov[1]*w.T
    return(-expectedUtility)

P = np.eye(5)
cons = ({'type':'eq', 'fun':lambda x: np.sum(x)-1})
bnds = tuple((0,1) for x in range(5))

postRetData = None
BLResData = []
for k in range(predStartN, len(pdata)):
    curDate = predDate[k - predStartN]
    sample = pdata.iloc[(k-windowSize-1):(k - 1)]
    eccessRet = sample[range(6)].sub(sample.interest, axis = 0)
    mktIndexData = sm.add_constant(eccessRet['hs300'])
    impliedCapmRet = np.zeros(5)
    histCampRet = np.zeros((5, len(sample)))
    for p in range(5):
        ticker = sample.columns[p]
        lrModel = sm.OLS(eccessRet[ticker], mktIndexData)
        lrFit = lrModel.fit()
        histCampRet[p] = lrFit.predict(mktIndexData)
        impliedCapmRet[p] = lrFit.predict(mktIndexData.mean(axis = 0))
    
    excessRetCov = np.mat(eccessRet.corr().ix[range(5), range(5)])
    Omega = np.mat(pd.DataFrame(np.array(forecastCov.loc[curDate]).reshape(forecastCovShape)).iloc[range(5), range(5)])
    Omega = np.mat(np.diag(np.diag(Omega)))
    Q = np.mat(np.array(predData.loc[curDate])).T
    M = ((tao*excessRetCov).I + P.T*Omega.I*P).I
    postRet = M*((tao*excessRetCov).I*np.mat(impliedCapmRet).T + P.T*Omega.I*Q)
    
    if postRetData is None:
        postRetData = pd.DataFrame(postRet.T)
    else:
        postRetData = postRetData.append(pd.DataFrame(postRet.T))
    
    retCov = (postRet, M)
    iniWeight = np.array([0.2]*5)
    optRes = sco.minimize(negPortfolioExpectedUtility, x0 = iniWeight, args = retCov, method = 'SLSQP', bounds = bnds, constraints = cons)
    BLWeight = np.round(optRes.get('x'), 5)
    BLPortfolioRet = np.array(pdata.loc[curDate][:5]).dot(BLWeight)
    BLRes = BLWeight.tolist()
    BLRes.append(BLPortfolioRet)
    BLResData.append(BLRes)
   

postRetData.set_index(pd.to_datetime(predDate), inplace = True)   
postRetData.columns = mdata.columns[:5].map(lambda x : x + '_post')

    
for index in mdata.columns[:5]:
    pd.merge(mdata[str(index)].to_frame(), postRetData[str(index) + '_post'].to_frame(),  left_index = True, right_index = True).plot()
```
