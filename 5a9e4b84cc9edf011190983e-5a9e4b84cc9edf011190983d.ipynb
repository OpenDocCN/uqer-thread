# 股价创n日新高（新低）买入（卖出）的策略

# 实验概述
基于以下策略：
- 当股票的收盘价大于前n日的最高价时，发出买入信号
- 当股票的收盘价小于前n日的最低价时，发出卖出信号
- 看不同n取值的效果

```python
import datetime
import pandas as pd
import numpy as np
import time


# 拿到所有股票的代码
sec_list = DataAPI.SecIDGet(partyID=u"",assetClass=u"E",ticker=u"",cnSpell=u"",field=u"",pandas="1")
sec_list = sec_list[sec_list.exchangeCD.isin(["XSHE", "XSHG"])]
sec_list = np.unique(sec_list.secID.values)

# 取最近一年的所有股票的行情数据
stock_price = DataAPI.MktEqudAdjGet(tradeDate=u"",secID=sec_list,ticker=u'',isOpen="",beginDate=u"20160302",endDate=u"20180302",field=[u"secID", 'ticker', 'tradeDate', 'highestPrice', 'lowestPrice', 'closePrice', 'openPrice'],pandas="1")
stock_price['tradeDate'] = stock_price['tradeDate'].apply(lambda x: x.replace("-", ""))

open_price = stock_price.pivot(index='tradeDate', columns='secID', values='openPrice')
```

```python
# 针对具体某个股票，计算前n日的最高、最低价格
def get_high_low(df, n):
    df.sort_values(by=['tradeDate'], ascending=True, inplace=True)
    df['newhigh'] = df['highestPrice'].shift(1)
    df['newlow'] = df['lowestPrice'].shift(1)
    df['nhigh'] = pd.rolling_max(df['newhigh'], window=n, min_periods=1)
    df['nlow'] = pd.rolling_min(df['newlow'], window=n, min_periods=1)
    return df

# 得到每个股票每天前N天的最高、最低价格
def get_highlow_n(n):
    global stock_price
    x_frame = stock_price.groupby(['secID']).apply(get_high_low, (n))
    return x_frame

# 判断是否达到买入/卖出条件
def get_order(order_frame):
    order_frame.dropna(inplace=True)
    order_frame['buy_status'] = order_frame['closePrice'] - order_frame['nhigh']
    order_frame['sell_status'] = order_frame['closePrice'] - order_frame['nlow']
    order_frame['buy_status'] = order_frame['buy_status'].apply(lambda x: 1 if x >0 else 0)
    order_frame['sell_status'] = order_frame['sell_status'].apply(lambda x: -1 if x < 0 else 0)
    order_frame['direction'] = order_frame['buy_status'] + order_frame['sell_status']
    order_frame = order_frame[order_frame.direction != 0]
    precise_frame = order_frame[['secID', 'tradeDate', 'direction']]
    return precise_frame

# 计算前n天最高、最低 买入/卖出策略的个股买卖状态
def calc_buy_sell(n):
    order_frame = get_highlow_n(n)
    final_frame = get_order(order_frame)
    return final_frame

signals = {}
for i in [5, 25, 50, 75, 100]:
    print "n=%s"%i
    x = calc_buy_sell(i).set_index('tradeDate')
    signals[i] = x
```

```python
##############################################################################
# 行情预加载函数
##############################################################################
from quartz.context.parameters import SimulationParameters
from quartz.backtest_tools import get_backtest_data


def preload_market_service(start, end, universe, benchmark, max_history_window=30, **kwargs):
    parameters = {
        'start': start,
        'end': end,
        'universe': universe,
        'benchmark': benchmark,
        'max_history_window': max_history_window
    }
    sim_parameters = SimulationParameters(**parameters)
    market_service = get_backtest_data(sim_parameters)
    market_service.rolling_load_daily_data(sim_parameters.trading_days)
    return market_service
```
