```python
# 转换stockcode代码为str类型，因为保存为CSV之后前面的0会消失
def code2str(x):
    x_str = str(x)
    while int(x)<100000:
        x = x*10
        x_str = '0' + x_str
    return x_str
```

```python
original_df = pd.read_csv('original_data.csv',encoding='GBK')
original_df['stockcode'] = original_df['stockcode'].apply(code2str)
# 转换tradeDate数据格式'%Y/%m/%d'成为提取因子可读取的'%Y-%m-%d'datetime形式
# original_df['tradeDate'] = original_df.tradeDate.apply(lambda x: dt.datetime.strptime(x,'%Y/%m/%d'))
```

```python
# 提取交易日期对应的可交易的ykcode用以提取可转债价格信息，可能会重复，因为会同时发布两个可转债，要去重
def get_ykcode(df):
    date_list = list(set(df.tradeDate.values))
    ykcode_list = []
    for date in date_list:
        ykcode = list(set(df.ykcode[df.tradeDate == date]))
        ykcode_list.append(ykcode)
    new_df = pd.DataFrame({'date':date_list,'ykcode_list':ykcode_list})
    return new_df

# 再次读取文件之后，是一个整体的str，转化为str构成的list
# 每个ykcode的长度为11，中间还有一个逗号和一个空格,首尾有[],但是缺少逗号和空格
def ykcodestr2strlist(x):
    return [x[1+i*11+i*2:1+(i+1)*11+i*2] for i in range(int(len(x)/13))]
```

```python
# 首次要运行，根据tradeDate每一天提取ykcode
ykcode_by_tradeDate_df=get_ykcode(original_df) 
ykcode_by_tradeDate_df.to_csv('ykcode_by_tradeDate.csv', encoding='gbk',index=False)
```

```python
# 再次运行时，从ykcode_by_tradeDate.csv中，ykcode会成为一整个字符串，要重新编排格式
ykcode_by_tradeDate_df = pd.read_csv('ykcode_by_tradeDate.csv', encoding='gbk')
ykcode_by_tradeDate_df['ykcode_list'] = ykcode_by_tradeDate_df.ykcode_list.apply(lambda x: ykcodestr2strlist(x))
```

```python
# 提取交易日期对应的可交易的stockcode，可能会重复，因为会同时发布两个可转债，要去重
def get_stockcode(df):
    date_list = list(set(df.tradeDate.values))
    stockcode_list = []
    for date in date_list:
        stockcode = list(set(df.stockcode[df.tradeDate == date]))
        stockcode_list.append(stockcode)
    new_df = pd.DataFrame({'date':date_list,'stockcode_list':stockcode_list})
    return new_df

# 从文件中读取stockcode_by_tradeDate，要把stockcode的一整个str转化为由str组成的list
# stockcode长度为2，仍旧有空格和逗号
def stockcodestr2list(x):
    return [x[1+i*6+i*2:1+(i+1)*6+i*2] for i in range(int(len(x)/8))]
```

```python
# 每次必须重新计算，否则提取不了一个一个的str组成list
stockcode_by_tradeDate_df=get_stockcode(original_df) 
stockcode_by_tradeDate_df.to_csv('stockcode_by_tradeDate.csv', encoding='gbk',index=False)
```

```python
# 从文件中读取stockcode
stockcode_by_tradeDate_df = pd.read_csv('stockcode_by_tradeDate.csv', encoding='gbk')
stockcode_by_tradeDate_df['stockcode_list'] = stockcode_by_tradeDate_df.stockcode_list.apply(lambda x: stockcodestr2list(x))
```

```python
# get_startend生成开始-结束日期，为季度数据提取做准备
# 生成标准日期格式再匹配
# 有可转债的中文名字可能会重复，但是ykcode不会重复
# 去掉销售日期，get_cov不重复地获取所有债券名单
# 要生成values进行布尔索引，如果直接选取一列，会产生series，不能对由values生成的array进行布尔索引 ！！！
def get_cov(df):
    new_df = df.drop(['tradeDate'],axis = 1)
    new_df = new_df.drop_duplicates() 
    new_df = new_df.reset_index(drop = True)
    return new_df

def get_startend(df):   
    new_df = get_cov(df)
    # 使用Series类型判断bool类型，再返回Series类型索取。生成的类型与被索取的对象要同时为Series或者array，否则无法匹配。
    new_df['start_date'] = new_df.apply(lambda x: min(df['tradeDate'][df.ykcode == x.ykcode]),axis = 1)
    new_df['end_date'] = new_df.apply(lambda x: max(df['tradeDate'][df.ykcode == x.ykcode]),axis = 1)
    return new_df
```

```python
# 当数据非三大基础季度数据，publishDate非财报发布日期时，所以将时间点转换成为时间段，本publishDate为intoDate，下一个publishDate为outDated
def point2period(df,pt_str = 'publishDate',new_str = '2022-01-01',key_str = 'ticker'):
    # 先将所有的数据按照ticker,pt升序排列
    df = df.sort_values(by = [key_str,pt_str],axis = 0,ascending = True)
    df['intoDate'] = list(df.publishDate.values)
    df['outDate'] = list(df.publishDate.values[1:])+ [new_str] 
    # 注意，因为intoDate和outDate有重叠，之后判断的时候要小于outdate！！！
    # 去掉交界处
    temp1 = df[key_str].values[1:]
    temp2 = df[key_str].values[:-1]
    df['ticker1'] = list(temp1 == temp2) + [False]
    df['outDate'] = df.apply(lambda x: new_str  if x.ticker1 == False else(x.outDate),axis = 1)
    df.drop(['ticker1'],inplace=True,axis = 1)
    return df
```

```python
def get_D_values(factor_list,stockcode_by_tradeDate_df,API_str = 'OneDayGet'):
    new_df = pd.DataFrame()
# 提取市值的API
    if API_str == 'MktEqudGet':
        for i in stockcode_by_tradeDate_df.index:
            date = stockcode_by_tradeDate_df.date.loc[i]
            stock_list = stockcode_by_tradeDate_df.stockcode_list.loc[i]
            temp_df = DataAPI.MktEqudGet(secID=u"",ticker=stock_list,
                                            tradeDate=date,beginDate=u"",endDate=u"",isOpen="",
                                            field=['ticker','tradeDate']+factor_list,pandas="1")
            new_df = new_df.append(temp_df,ignore_index=True)
# 提取基础因子的API
    if API_str == 'OneDayGet':
        for i in stockcode_by_tradeDate_df.index:
            date = stockcode_by_tradeDate_df.date.loc[i]
            stock_list = stockcode_by_tradeDate_df.stockcode_list.loc[i]
            temp_df = DataAPI.MktStockFactorsOneDayGet(tradeDate=date,secID=u"",
                                                       ticker=stock_list,
                                                       field=['ticker','tradeDate']+factor_list,pandas="1")
            new_df = new_df.append(temp_df,ignore_index=True)
    if API_str == 'BondPremiumGet':
         for i in stockcode_by_tradeDate_df.index:
            date = stockcode_by_tradeDate_df.date.loc[i]
            SecID_list = stockcode_by_tradeDate_df.ykcode_list.loc[i]
            temp_df = DataAPI.MktConsBondPremiumGet(SecID=SecID_list,tickerBond=u"",
                                              beginDate=date,endDate=date,
                                              field=['SecID','tradeDate']+factor_list,pandas="1")
            new_df = new_df.append(temp_df,ignore_index=True)
    return new_df
```

```python
# 往前多提取N个日，否则前面N-?个交易日就没有数据了
# 构建新的ykcode_by_tradeDate_df，先按照时间排序该表，然后把每一行提前N个交易日合并后去重生成新的表
# code_ahead是把所有code提前了N天，code是原始的对应日期
# 针对所给交易日不是连续的还没有写判断规则！！！
def set_date_ahead(df,N,code_str):
    df.reset_index(inplace=True,drop=True)
    datestart = dt.datetime.strptime(df.date[0],'%Y-%m-%d')
    date_list = []
    for i in range(N):
        new_date = datestart + dt.timedelta(days=-N+i)
        date_list.append(new_date.strftime('%Y-%m-%d'))
    # print(date_list)
    date_list = date_list+df.date.values.tolist() # 直接values没有办法list加上去，series没有办法，会直接把str加长
    new_df = pd.DataFrame(columns=['date','code_ahead','code'])
    new_df['date'] = date_list
    # print(date_list)
    new_df['code_ahead'][:-N] = df[code_str]
    new_df['code'][N:] = df[code_str]
    # print(new_df.index)
    
    def fun_merge(x): # 把两个code合并在一起
        if isinstance(x.code, float): # 判断数据类型，因为nan是float，不是list，不能直接相加
            return x.code_ahead
        elif isinstance(x.code_ahead, float):
            return x.code
        else:
            return list(set(x.code+x.code_ahead))
    
    new_df['new_code'] = new_df.apply(lambda x: fun_merge(x), axis = 1)
    new_df.drop(['code','code_ahead'],inplace = True,axis = 1)
    return new_df
```

# 按照intoDate和outDate添加行业数据

```python
def add_industry(new_df,filename,ind_type):
    industry_df = pd.read_csv(filename,encoding='gbk')
    # industry_df 行业标签
    indID = []

    new_df.reset_index(drop=True,inplace=True)
    for i in new_df.index:
        temp1 = industry_df.ticker.values==new_df.stockcode.values[i]
        temp2 = industry_df.intoDate <= new_df.tradeDate[i]
        temp3 = industry_df.outDate > new_df.tradeDate[i]
        # 存在提取不到时间的
        temp4 = industry_df.industryID1[temp1&temp2&temp3]
        if len(temp4) == 0:
            indID.append(np.NAN)
        else:
            indID.append(temp4.values[0])
    new_df[ind_type] = indID
    return new_df

def add_bond(new_df,filename,ind_type):
    industry_df = pd.read_csv(filename,encoding='gbk')
    # industry_df 行业标签
    indID = []

    new_df.reset_index(drop=True,inplace=True)
    for i in new_df.index:
        temp1 = industry_df.secID.values==new_df.ykcode.values[i]
        temp2 = industry_df.intoDate <= new_df.tradeDate[i]
        temp3 = industry_df.outDate > new_df.tradeDate[i]
        # 存在提取不到时间的
        temp4 = industry_df.rating[temp1&temp2&temp3]
        if len(temp4) == 0:
            indID.append(np.NAN)
        else:
            indID.append(temp4.values[0])
    new_df[ind_type] = indID
    return new_df
```
