### 0.1 欧式期权定价模型 - Black-Scholes-Merton
---
- 欧式定价模型是最基础的模型啦，但也是应用频率最多的模型，它的定价公式是Black-Scholes模型，很多网站都有它的资料，在这里就用Python的code来实现了。对于高校课程来说，它是比较基础的期权定价模型，但对散户朋友来说，它就是一坨浆糊 =、= 确实，定价算得准也不见得抓得住波段行情，行情看不准，期权投资容易变成无本之木、无水之源。总之，这里先把BS模型码出code来。
- 伴随着定价模型产生的，就是希腊字母和隐含波动率。详细深奥的内容就不去探讨了，比如“Monte-Carlo定价时怎样求Delta”，基础的BS模型求Delta还是比较直观的，就是数学上的偏微分概念。另外，求隐含波动率IV的方法也有多种，这里是比较容易想到的“二分法”。

```python
"""欧式期权定价公式，也就是Black-Scholes模型"""
def bs_model(A,S,K,r,T,sigma):
    S=float(S); K=float(K);
    if A==1:#认购期权
        if T==0:
            p = max(0 , S - K)
        elif T>0:
            d = (np.log(S / K) + T *  0.5 * sigma**2) / (sigma * np.sqrt(T));
            p = np.exp(-1*r*T) * ( S* stats.norm.cdf(d) -  K * stats.norm.cdf(d - sigma * np.sqrt(T)));
        else:
            raise ValueError('In bs_model, T is < 0 !')
    elif A==0:#认沽期权
        if T==0:
            p = max(0 , K - S)
        elif T>0:
            d = (np.log(S / K) + T *  0.5 * sigma**2) / (sigma * np.sqrt(T));
            p = np.exp(-1*r*T) * ( K*stats.norm.cdf(sigma*np.sqrt(T)-d) - S* stats.norm.cdf(-1*d));
        else:
            raise ValueError('In bs_model, T is < 0 !')
    else:#确定参数符合规则
        raise ValueError('In bs_model, A is neither 1 nor 0 !')
    return p


"""用数值法求偏微分，也就是期权的Delta。实际上Delta用这种方法算还比较准，但其他几个希腊字母就未必很准"""
def get_delta(A,S,K,r,T,sigma,func=bs_model, *args):
    epsilon = 1e-2
    dS = K*epsilon
    delta =(func(A,S+dS,K,r,T,sigma,*args) - 
        func(A,S-dS,K,r,T,sigma,*args)) / (2*dS)
    return delta


"""二分法计算隐含波动率，到期日当天IV记作None，计算IV失败则记作-0.001"""
def get_IV(A,S,K,r,T,option_price,func=bs_model,*args):
    if T<0.1/365:#期权到期日，不计算IV
        return None
    
    iv_1=0.001; step=1.0;
    iv_2=iv_1+step;
    
    price1=func(A,S,K,r,T,iv_1,*args)
    price2=func(A,S,K,r,T,iv_2,*args)
    diff1=abs(option_price-price1)
    diff2=abs(option_price-price2)
    diff=min(diff1,diff2)
    if price1>=option_price: #期权最新价格太过便宜，实际上IV就是0.0
        return 0.0
    while price2<=option_price: #二分法要有合理的初始起点
        if iv_2>2.0: 
            return np.nan
        iv_2+=step
        price2=func(A,S,K,r,T,iv_2,*args)
    
    # 二分法的while循环
    n=0; n_limit=100;
    while n<n_limit and diff>max(0.00002,option_price*1e-6):
        iv_mid=0.5*(iv_1+iv_2)
        price_mid=func(A,S,K,r,T,iv_mid,*args)
        if price_mid>option_price:
            iv_2=iv_mid
        elif price_mid<option_price:
            iv_1=iv_mid
        else:
            return iv_mid
        n+=1
        price1=func(A,S,K,r,T,iv_1,*args)
        price2=func(A,S,K,r,T,iv_2,*args)
        diff1=abs(option_price-price1)
        diff2=abs(option_price-price2)
        diff=min(diff1,diff2)
        #print option_price,price_mid
    
    if n>=n_limit:
        return -0.001
    if diff1<diff2:
        return iv_1
    else:
        return iv_2
    
#print bs_model(1,2981 ,3000,0,85/365.0,0.2)
#print get_delta(1,2981 ,3000,0,85/365.0,0.2)
```

### 0.2 期权历史数据的小工具，T型报价表回放，波动率曲线画图
---
- 本文介绍一个基于优矿数据和Python的期权历史数据小工具，实际上就是一个Python类。
- 它具有1个__init__函数加载用户关心的所有期权数据，和4个计算函数来展现用户所关心的具体功能。（~这4个计算函数相互并联，没有相关依赖关系~）
- 它能提供历史每一个交易日收盘时刻的期权数据，相信足够满足多数投资者朋友的数据需求。

```python
class Option_Engine(object):
    def __init__(self,beginDate=u"20170901",endDate=u"20180301",varTicker="510050",exchangeCD=u"XSHG",target_YM=None):
        # 关于参数，前四个参数必须准确填写，交易所的英文简称请参考优矿Uqer的文档；而target_YM有助于加快隐含波动率的计算步骤
        print "Now, initializing..."
        
        # 抽取期权合约日行情，构成“总表”dfTotal的一部分内容
        field="tradeDate,ticker,secShortName,"+\
            "openPrice,highestPrice,lowestPrice,closePrice,"\
            "turnoverVol,turnoverValue,openInt"
        df=DataAPI.MktOptdGet(beginDate=beginDate,endDate=endDate,exchangeCD=exchangeCD,field=field)
        
        # 对不同交易所的期权命名规则，来分辨认购的"C"字母和认沽的"P"字母
        if exchangeCD in ["XSHG","XSHE"]:
            df["direction"]=df["ticker"].map(lambda x:x[6])
        elif exchangeCD in ["CCFX","XDCE"]:
            df["direction"]=df["ticker"].map(lambda x:x.split("-")[1])
        elif exchangeCD in ["XSGE","XZCE"]:
            def func(x):
                if x.rfind("C")>x.rfind("P"):
                    direction="C"
                else:
                    direction="P"
                len_StrikePrice=len(x.split(direction)[-1])
                varticker=x[:-1*(len_StrikePrice+1)]
                return (varticker,direction)
            df["direction"]=df["ticker"].map(lambda x:func(x)[1])
        else:
            raise ValueError(
                u"在__init__函数中，输入的exchangeCD参数，导致查询交易所合约的期权到期日，\
                无法找到对应交易所。这个情况可能是由于新的期权出现在了新一家交易所，请用户朋友在这里自行校对，见谅"
            )
        df=df.set_index(["tradeDate","ticker"])
        print "loading option-data finished"
        #print len(df)
        
        # 对不同交易所抽取标的物合约信息，对沪深证交所需要抽出复权日行情，对其他交易所需要抽出合约信息，之后汇聚成总表dfTotal
        if exchangeCD in ["XSHG","XSHE"]:
            field="tradeDate,ticker,strikePrice,lastTradeDate,varTicker"
            df1=DataAPI.OptDpoGet(varTicker=varTicker,beginDate=beginDate,endDate=endDate,field=field)
            df1=df1.set_index(["tradeDate","ticker"])
            dfTotal=pd.concat([df,df1],join="inner",axis=1)
            
        elif exchangeCD in ["CCFX","XDCE","XSGE","XZCE"]:
            field="tickerSymbol,strikePrice,lastTradeDate,varTicker"
            df1=DataAPI.OptGet(varticker=varTicker,contractStatus=u"",exchangeCD=exchangeCD,field=field)
            df1=df1.rename(columns={"tickerSymbol":"ticker"}).set_index("ticker")
            df=df.reset_index().set_index("ticker")
            strikeList=[]; lastTradeList=[];
            for i in range(len(df)):
                ind=df.index.values[i]
                if ind in df1.index.unique():
                    strikeList.append(df1["strikePrice"][ind])
                    lastTradeList.append(df1["lastTradeDate"][ind])
                else:
                    strikeList.append(None)
                    lastTradeList.append(None)
            dfTotal=copy.deepcopy(df)
            dfTotal["strikePrice"]=strikeList
            dfTotal["lastTradeDate"]=lastTradeList
            dfTotal=dfTotal.dropna()
            
        else:
            raise ValueError(
                u"在__init__函数中，输入的exchangeCD参数，导致查询交易所合约的期权到期日，\
                无法找到对应交易所。这个情况可能是由于新的期权出现在了新一家交易所，请用户朋友在这里自行校对，见谅"
            )
        
        dfTotal=dfTotal.reset_index()
        dfTotal["varticker"]=df1["varTicker"][0]
        dfTotal["td"]=dfTotal["tradeDate"].map(lambda x:dt.datetime.strptime(x,"%Y-%m-%d"))
        dfTotal["ld"]=dfTotal["lastTradeDate"].map(lambda x:dt.datetime.strptime(x,"%Y-%m-%d"))
        dfTotal["remainingDays"]=(dfTotal["ld"]-dfTotal["td"]).map(lambda x:x.days)
        del dfTotal["td"],dfTotal["ld"]
        dfTotal["lastTradeYM"]=dfTotal["lastTradeDate"].map(lambda x:x[0:7])
        if target_YM and len(target_YM)>0:  dfTotal=dfTotal[ dfTotal["lastTradeYM"]==target_YM ]
        dfTotal=dfTotal.set_index("tradeDate")
        print "loading contract-information finished"
        #print dfTotal

        # 对不同交易所抽取标的物合约的日行情，并计算历史波动率30HV和60HV
        field="tradeDate,ticker,secShortName,closePrice"
        if exchangeCD in ["XSHG","XSHE"]:
            dfVar=DataAPI.MktFunddGet(ticker=dfTotal["varticker"][0],beginDate=beginDate,endDate=endDate,field=field)
        elif exchangeCD in ["XSGE","XDCE","XZCE"]:
            dfVar=DataAPI.MktFutdGet(ticker=dfTotal["varticker"][0],beginDate=beginDate,endDate=endDate,field=field)
        elif exchangeCD in ["CCFX"]:
            field="tradeDate,ticker,secShortName,closeIndex"
            dfVar=DataAPI.MktIdxdGet(ticker=dfTotal["varticker"][0],beginDate=beginDate,endDate=endDate,field=field)
            dfVar=dfVar.rename(columns={"closeIndex":"closePrice"})
        else:
            raise ValueError(
                u"在__init__函数中，输入的exchangeCD参数，导致查询标的物日行情失败。目前仅支持沪深证券交易所，中金所和三大商品交易所"
            )
        dfVar["var_ret"]=dfVar["closePrice"]/dfVar["closePrice"].shift(1)-1
        dfVar["var_30HV"]=dfVar["var_ret"].rolling(window=30).std()*(252)**0.5
        dfVar["var_60HV"]=dfVar["var_ret"].rolling(window=60).std()*(252)**0.5
        dfVar=dfVar.set_index("tradeDate")
        
        closeVar=[]
        var_30HV=[]
        var_60HV=[]
        for td in dfTotal.index.values:
            closeVar.append(dfVar["closePrice"][td])
            var_30HV.append(dfVar["var_30HV"][td])
            var_60HV.append(dfVar["var_60HV"][td])
        dfTotal["closeVar"]=closeVar
        dfTotal["var_30HV"]=var_30HV
        dfTotal["var_60HV"]=var_60HV
        print "loading fund-or-futures finished"
        #print dfTotal

        # 最耗时的步骤，计算每一天每一张合约的隐含波动率IV
        closeIV=[]
        RISKFREE=0.0
        direction=dfTotal["direction"].tolist()
        strikePrice=dfTotal["strikePrice"].tolist()
        remainingDays=dfTotal["remainingDays"].tolist()
        closePrice=dfTotal["closePrice"].tolist()
        print "\n","Start calculating implied-volatility:"
        for i in range(len(dfTotal)):
            S=closeVar[i]
            K=strikePrice[i]
            r=RISKFREE
            T=remainingDays[i]/365.0
            option_price=closePrice[i]
            if direction[i]=="C":
                A=1
            elif direction[i]=="P":
                A=0
            else:
                raise ValueError("direction is not C-nor-P, while get_IV()")
            IV=get_IV(A,S,K,r,T,option_price)
            closeIV.append(IV)
            
            percent_lag=int((i-1)/float(len(dfTotal))*100.0)
            percent=int(i/float(len(dfTotal))*100.0)
            if percent_lag<percent:
                print "Now have calculated %s IV-data points(%s percent), total-length is %s" %(i,percent,len(dfTotal))

        dfTotal["closeIV"]=closeIV
        self.dfTotal=dfTotal
        print "Calculating Implied-Volatility finished"
        #print dfTotal
     
    
    """"
    这个函数用于画出历史上某一天，某张合约的隐含波动率曲线，和T型报价表信息，即dfCall和dfPut。
    值得注意的是，target_YM参数用来选出特定到期日的期权，而且必须注意它的含义是期权最后交易日所处的年份和月份，
    如SR1801合约到期日在2017年11月，但是m1801合约到期日在2017年12月，若是金融期权则多数是名称当月，请自行留意。
    """
    def plot_IV_daily(self,target_td="2018-02-09",target_YM="2018-03",flag_plot=True):
        # 检验日期是否符合规则
        tradeDate_list=sorted(np.unique(self.dfTotal.index.values))
        if target_td not in tradeDate_list:
            raise ValueError(u"plot_IV_daily函数中,target_td参数有误，请输入交易日期 %Y-%m-%d")
            
        # 切片出对应日期、对应期权合约的数据，分成Call和Put两边
        dfTarget=self.dfTotal.loc[target_td,:]
        dfTarget=dfTarget[ dfTarget["lastTradeYM"]==target_YM ]
        varticker=dfTarget["varticker"].values[0]
        dfCall=dfTarget[dfTarget["direction"]=="C"].sort_values(by="strikePrice")
        dfPut=dfTarget[dfTarget["direction"]=="P"].sort_values(by="strikePrice")
        
        ind_yesterday=tradeDate_list.index(target_td)-1
        if ind_yesterday>=0:
            dfYester=self.dfTotal.loc[tradeDate_list[ind_yesterday],:]
            dfYester=dfYester[ dfYester["lastTradeYM"]==target_YM ]
            dfCall_lag1=dfYester[dfYester["direction"]=="C"].sort_values(by="strikePrice")
            dfPut_lag1=dfYester[dfYester["direction"]=="P"].sort_values(by="strikePrice")
            
            strike_list=dfCall["strikePrice"].tolist()
            strikeYester_list=dfCall_lag1["strikePrice"].tolist()
            dfCall["openInt_diff"]=0
            for K in strike_list:
                if K in strikeYester_list:
                    ind=strike_list.index(K)
                    yester_openInt=(dfCall_lag1[dfCall_lag1["strikePrice"]==K])["openInt"].values[0]
                    dfCall.ix[ind,"openInt_diff"]=dfCall.ix[ind,"openInt"]-yester_openInt
                else:
                    dfCall.ix[ind,"openInt_diff"]=None
                    
            strike_list=dfPut["strikePrice"].tolist()
            strikeYester_list=dfPut_lag1["strikePrice"].tolist()
            dfPut["openInt_diff"]=0
            for K in strike_list:
                if K in strikeYester_list:
                    ind=strike_list.index(K)
                    yester_openInt=(dfPut_lag1[dfPut_lag1["strikePrice"]==K])["openInt"].values[0]
                    dfPut.ix[ind,"openInt_diff"]=dfPut.ix[ind,"openInt"]-yester_openInt
                else:
                    dfPut.ix[ind,"openInt_diff"]=None
        else:
            dfCall["openInt_diff"]=None
            dfPut["openInt_diff"]=None
                
        # 画图
        dfMerge=pd.DataFrame()
        dfMerge["strikePrice"]=dfCall["strikePrice"].values
        dfMerge["call_IV"]=dfCall["closeIV"].values
        dfMerge["put_IV"]=dfPut["closeIV"].values
        if flag_plot:
            dfMerge.plot(
                x="strikePrice",y=["call_IV","put_IV"],figsize=(16,6),style=["r-","b-"],
                title="IV-curve:     %s_option   lastTrade_YM:%s   on_date:%s"%(varticker,target_YM,target_td)
            )
        columns=["secShortName","ticker","direction","strikePrice","turnoverVol","openInt","openInt_diff","closePrice","closeIV"]
        return dfCall[columns],dfPut[columns]
    
    def plot_option_kbars(self,ticker,beg_date="2018-01-01",end_date="2019-01-01",flag_plot=True):
        flag=( 
            (self.dfTotal.index.values>=beg_date)&
            (self.dfTotal.index.values<=end_date)&
            (self.dfTotal["ticker"]==ticker)
        )
        columns=["openPrice","highestPrice","lowestPrice","closePrice","closeVar"]
        df_kbars=self.dfTotal[flag][columns]
        if flag_plot:
            print "option_kBars:      varTicker_%s(left-Y)    option-ticker_%s(right-Y)    grid_on: right-Y" \
                %( (self.dfTotal[flag])["varticker"][0], ticker)
            ax=df_kbars["closeVar"].plot(figsize=(16,6))
            ax1=ax.twinx()
            candlestick2_ohlc(#这个函数来自matplotlib.finance，当时matplot版本是1.5.1，新版里它位置有改变
                ax1,
                df_kbars["openPrice"],
                df_kbars["highestPrice"],
                df_kbars["lowestPrice"],
                df_kbars["closePrice"],
                colorup='r', colordown='g',
                width=0.4, alpha=1.0
            )
            ax1.grid(axis="both",color='k', linestyle='--', linewidth=0.5,alpha=0.25)
        return df_kbars    

    
    """"
    这个函数用于画出历史上某一段时间，某标的物不同行权价合约们整体，在平值附近的隐含波动率，的阶段性走势图。
    同样要注意，target_YM参数用来选出特定到期日的期权，而且必须注意它的含义是期权最后交易日所处的年份和月份，
    如SR1801合约到期日在2017年11月，但是m1801合约到期日在2017年12月，若是金融期权则多数是名称当月，请自行留意。
    另外，price_gap这个参数很粗略地定义了“什么是平值附近”，具体用法需看下文的code
    """
    def plot_IV(self,beg_date="2018-01-01",end_date="2019-01-01",target_YM="2018-03",flag_plot=True,price_gap=0.15):   
        flag=( 
            (self.dfTotal.index.values>=beg_date)&
            (self.dfTotal.index.values<=end_date)&
            (self.dfTotal["lastTradeYM"]==target_YM)
        )
        dfTotal_YM=self.dfTotal[flag]
        #print dfTotal_YM

        tradeDate_list=[]
        call_iv_list=[]
        put_iv_list=[]
        var30HV_list=[]
        var60HV_list=[]
        for date in dfTotal_YM.index.values:
            dfCut=dfTotal_YM.loc[date,:]
            closeVar=dfCut["closeVar"].values[0]
            tradeDate_list.append(date)
            var30HV_list.append(dfCut["var_30HV"].values[0])
            var60HV_list.append(dfCut["var_60HV"].values[0])

            flag_call= (dfCut["direction"]=="C") & (dfCut["strikePrice"]>=closeVar-price_gap) & (dfCut["strikePrice"]<=closeVar+price_gap)
            iv_call=sorted(dfCut[flag_call]["closeIV"].tolist())
            if iv_call[0]>np.mean(iv_call)-np.std(iv_call):
                call_iv_list.append(iv_call[0])
            else:
                call_iv_list.append(np.mean(iv_call))

            flag_put= (dfCut["direction"]=="P") & (dfCut["strikePrice"]>=closeVar-price_gap) & (dfCut["strikePrice"]<=closeVar+price_gap)
            iv_put=sorted(dfCut[flag_put]["closeIV"].tolist())
            if iv_put[0]>np.mean(iv_put)-np.std(iv_put):
                put_iv_list.append(iv_put[0])
            else:
                put_iv_list.append(np.mean(iv_put))

        dfIV=pd.DataFrame()
        dfIV["tradeDate"]=tradeDate_list
        dfIV["call_IV"]=call_iv_list
        dfIV["put_IV"]=put_iv_list
        dfIV["30HV"]=var30HV_list
        dfIV["60HV"]=var60HV_list
        dfIV=dfIV.set_index("tradeDate")
        if flag_plot:
            ax=dfIV.plot(figsize=(16,6),style=["r-","b-","y--","g--"])
        return dfIV    
    
    
    """"
    这个函数只能用于上海交易所和深圳交易所，画出历史时间全集，某ETF期权合约们整体在平值附近的隐含波动率的阶段性走势图。
    大体上就是一个for循环不断调用plot_IV函数，因此参数用法也相同。
    """
    def plot_IV_total(self,flag_plot=True,figsize=(32,8),price_gap=0.15):
        tradeDate_list=sorted(np.unique(self.dfTotal.index.tolist()))
        beg_index=0; end_index=beg_index+1;
        dfList=[]
        while end_index<len(tradeDate_list):
            flag = tradeDate_list[end_index-1][-2:]<="10" and tradeDate_list[end_index][-2:]>"10"
            if flag:
                beginDate=tradeDate_list[beg_index]
                endDate=tradeDate_list[end_index-1]
                target_YM=endDate[0:7]
                dfIV=engine.plot_IV(beg_date=beginDate,end_date=endDate,target_YM=target_YM,flag_plot=False,price_gap=price_gap)
                dfList.append(dfIV)
                beg_index=end_index
                end_index=beg_index+1
            else:
                end_index+=1
        dfIV=engine.plot_IV(beg_date=beginDate,end_date=endDate,target_YM=target_YM,flag_plot=False,price_gap=price_gap)
        dfList.append(dfIV)
        dfIV_total=pd.concat(dfList,axis=0)
        
        if flag_plot:
            #dfIV_total=dfIV_total.dropna()
            ax=dfIV_total.plot(figsize=figsize,style=["r-","b-","y--","g--"])
        return dfIV_total
```

### 1.0 创建结构体，下载历史数据总集

- (1) 这个步骤耗时是最长的，运行过程中会print进度提示
- (2) 4个必须输入的参数，用于明确：历史数据的开始时间、结束时间，标的物ticker，和具体所在的交易所

### 1.3 从历史数据总集中抽取某段时间的历史数据，并整合为隐含波动率IV与历史波动率HV的走势图
---
- 这个函数运行所消耗的时间中等，不是很短也不会很长
- 4个参数，前三个参数用于指定：开始时间、结束时间、期权的“到期日所在的年份和月份”
- 最后一个参数是price_gap，用来很粗略地规定“何为平值附近”。具体地说，比如50ETF期权最近每档行权价间距是0.1，那么price_gap填0.1到0.3左右都是比较合适的；
但如果对于上期所的黄金期货期权来说，每档行权价是4元，那么price_gap应该设定在2.0到12.0左右；
- 隐含波动率是某些研究员和交易员津津乐道的话题，不过实战中即使不是弄得很明白，照样也能赚钱，因为看懂标的物的涨跌方向、行情节奏更重要

```python
dfIV=engine.plot_IV(beg_date="2020-08-01",end_date="2020-08-31",target_YM="2020-09",price_gap=0.15)
```

### 1.4 对历史数据总集做整理，画出全时间范围的隐含波动率IV和历史波动率HV走势图
---
- 这个函数运行时间比较长，但通常不超过10分钟，即使重复运行也尚可接受的

- 它只有一个price_gap参数，含义与上文plot_IV函数里的意义相同，因为这个函数就是通过重复运行plot_IV函数实现的
- 这个函数主要是用来验证某些学院派的观点，如“长期而言，隐含波动率总是围绕历史波动率上下浮动”
- 对于其他交易所，不可以调用此函数的情况，请继续用上文的plot_IV函数，多次调用多画几张图。比如，不可以将“豆粕期权波动率”拼接在一起，因为这里面涉及到“标的物到底是谁”的问题。比如，豆粕m1801合约的期权，其标的物是m1801期货合约，而m1805期权，其标的物合约是m1805期货合约。为什么不可以拼接在一起呢？因为同一段时间内，期货的m1801和m1805两张合约由于物流、天气、产区等基本面的原因，可以走出“背离”的行情，所以它们的期权严格意义上也不可以一同比较。如果觉得此事有蹊跷，可以拍打脑壳咱们再讨论讨论 >_<
