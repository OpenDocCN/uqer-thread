# 股票暴涨的频繁模式识别

```python
import numpy as np
import pandas as pd
import datetime
import time
```

```python
##股票选取
def chs_factor(ticker = '000005',begin = '20010101',end = '20170201'):
    dataday = DataAPI.MktEqudGet(tradeDate=u"",secID=u"",ticker=ticker,beginDate= begin,endDate= end,isOpen="1",field=u'',pandas="1")
    return dataday
```

```python
##股票特征处理
def chs_ticker(dataday,n = 4):
    k = list(pd.qcut(dataday,n,labels = [str(x) for x in range(1,n+1)]))
    l = k[0]
    for i in k:
        l = l+i
    l = l[1:]
    return l
```

```python
##分位数提取
def percent(p,n):
    perti = []
    for i in np.linspace(0,100,n+1)[1:-1]:
        pt = np.percentile(p,i)
        perti.append(pt)
    return perti
```

```python
##序列生成函数
def n_list(glist,x = 2):
    relist = []
    for k in range(0,len(glist)-x+1):
        relist.append(glist[k:k+x])
    return relist
```

```python
##暴涨模式
def dz_parn(dazhang,n = 4):
    dz_pattern = []
    for i in dazhang:
        if i > 5:
            i_add = [x for x in range(i-n+1,i+1)]
            t = list(dataday.ix[i_add]['label'])
            f = t[0]
            for i in t[1:]:
                f = f+i
            dz_pattern.append(f)
    return dz_pattern
```

```python
##ck序列与置信度生成函数
def ck(frq,x,n=4):
    klist = []
    frq_k = []
    sup_k = []
    bay_s = []
    re_list = n_list(l,x-1)
    ren_list = n_list(l,x)
    total_num = len(ren_list)
    for i in frq:
        b = np.float(re_list.count(i))
        for j in [str(x) for x in range(1,n+1)]:
            k = i+j
            p = ren_list.count(k)
            if p/np.float(total_num)>0.006:
                if (p/np.float(total_num))>(1/np.float(n**x+2**x)):
                    frq_k.append(k)
                    sup_k.append(p/np.float(total_num))
                    bay_s.append(p/b)
    return frq_k,sup_k,bay_s
```

```python
##同时满足暴涨模式与频繁模式
def chs_dz(frq_):
    choose_list = []
    for choose in frq_:
        if choose in dz_parn(dazhang):
            choose_list.append(choose)
    return choose_list
```

```python
##暴涨模式的整体占比
def dz_rate(choose_list):
    dzrate = []
    for parn in choose_list:
        relist = n_list(l,4)
        dzrate.append(dz_pattern.count(parn)/np.float(relist.count(parn)))
    return dzrate
```

```python
tick = list(DataAPI.MktEqudGet(tradeDate=u"20170103",secID=u"",ticker=u'',beginDate=u'',endDate= u'',isOpen="1",field=u'',pandas="1")['ticker'])
```

```python
now = time.time()
for ticker in tick[10:20]:
    dataday = chs_factor(ticker)
    dataday['label'] = pd.qcut(dataday['chgPct'],4,labels = ['1','2','3','4'])
    l = chs_ticker(dataday['chgPct'])
    dazhang = dataday[dataday['chgPct']>0.05].index
    frq = list(set(n_list(l,2)))
    frq = ck(frq,3)[0]
    if len(ck(frq,4)[0]) != 0:
        frq_,sup,cond = ck(frq,4)
        ku = pd.DataFrame([frq_,sup,cond],index = ['frq','sup','con']).T
        choose_list = chs_dz(frq_)
        ku.index = list(ku.frq)
        ku = ku.ix[choose_list]
        ku['dzrate'] = dz_rate(choose_list)
        ku.index = range(len(ku))
    print ticker,'\n',ku
time.time()-now
```
