```python
import datetime
import pandas as pd
def datecaculate(today,steps):  
    #today=datetime.date.today()
    if datetime.date.isoweekday(today)>steps%5:
        datecaculate=(today-datetime.timedelta(7*(steps/5))-datetime.timedelta(steps%5)).strftime('%Y-%m-%d')
    else:
        datecaculate=(today-datetime.timedelta(7*(steps/5))-datetime.timedelta(2)-datetime.timedelta(steps%5)).strftime('%Y-%m-%d')
    return datecaculate

import pandas as pd #get the stocks without st and new ones#
def zhengquanchi(yesterday,date_40):
    gpc=(DataAPI.MktEqudGet(tradeDate=yesterday,isOpen="1",
    field=u"secID,ticker,tradeDate,secShortName,closePrice,turnoverRate,negMarketValue",pandas="1"))
    gpc = gpc[[True if x[0] in '036' else False for x in gpc.secID]]   #secID都是以0，3，6打头的
    gpc = gpc[[False if 'ST' in x else True for x in gpc.secShortName]]   # 剔除ST
    ssrq=DataAPI.SecIDGet(partyID=u"",assetClass=u"",ticker=gpc.ticker,cnSpell=u"",field=u"secShortName,listDate",pandas="1")
    gpc=pd.merge(gpc,ssrq,right_on='secShortName',left_on='secShortName')
    gpc=gpc[gpc['listDate']<date_40]
    return gpc['secID']
today=datetime.date.today()
yesterday=datecaculate(today,1)
date_40=datecaculate(today,40)
zhengquanchi(yesterday,date_40)
```
