```python
#3.决策树 (Decision Tree)
#Import Library
#导入库
#Import other necessary libraries like pandas, numpy...
#导入其他必须的包如pandas，numpy
#以上是无关紧要的话

from sklearn import tree
#导入sklearn中的tree包
#Assumed you have, X (predictor) and Y (target) for training data set and x_test(predictor) of test_dataset
#假设你已经拥有训练集和测试集，也就是你已经有大象了

# Create tree object 
#创建模型
model = tree.DecisionTreeClassifier(criterion='gini') # for classification, here you can change the algorithm as gini or entropy (information gain) by default it is gini  
#tree.DecisionTreeClassifier()是决策树模型的类，将model定义成这个类
#criterion='gini'是决策树种的一个参数，你可以更改成entropy或者gini，这个是决策树的分类方法。
#模型的名字DecisionTreeClassifier，从Classifier可以看出这个是分类算法
#这个就是造了一个冰箱

# model = tree.DecisionTreeRegressor() for regression
#如果是DecisionTreeRegressor，Regressor是回归算法
# Train the model using the training sets and check score
#利用训练集数据训练模型，并检查分数
model.fit(X, y)
#sklearn所有的模型几乎都有一个fit，就是训练模型，X是训练集的自变量，y是因变量
#这步就是把大象放进冰箱里

model.score(X, y)
#score函数可以检查对训练集分类的效果，分数越高效果越好

#Predict Output
predicted= model.predict(x_test)
#模型训练好以后，就可以用来预测了，基本所有的模型都可以用predict方法来预测，参数是测试集的自变量。
#这一步实际上就是把冰箱门关上了
```

上述的例子是最基本的流程，完全符合大象放进冰箱里的步骤，简单不简单？还要啥自行车。

### 问题2：利用机器学习构建量化模型总共分几步：
>答：四步
第一步，构建变量
第二步，构造模型
第三步，把数据放进模型
第四步，使用训练好的模型进行预测或分类

下面举一个基础的机器学习例子，看一下如何利用代码实现我们的想法。数据集采用sklearn自带的鸢尾花数据。


```python
from sklearn import datasets
from sklearn import tree
from sklearn.cross_validation import train_test_split

data=datasets.load_iris()['data']
target=datasets.load_iris()['target']
#导入数据，data是鸢尾花的花萼宽度长度、花瓣长度宽度，target是该鸢尾花的品种，有0,1,2三种类型

trainx,testx,trainy,testy=train_test_split(data,target,test_size=0.2)
#造一个大象，将源数据拆分成训练集和测试集，测试集是总数据的20%

clf=tree.DecisionTreeClassifier()
#创建一个名为clf的冰箱，创建一个模型，参数全部用默认

clf.fit(trainx,trainy)
#把大象塞进冰箱，训练这个模型

print '预测值：',clf.predict(testx)
#看看训练完的冰箱的工作能力

print '实际值：',testy
#看看实际的值

print '预测准确率：%2.2f%%'%(clf.score(testx,testy)*100)
#输出准确率

#由于train_test_split是随机分配训练和测试集，所以每次运行的结果会有所不同
```

上述代码完成了一个最基本的机器学习的流程，我构建了一个最基本的决策树模型，该模型的分类性能是通过鸢尾花的花萼长宽、花瓣长宽来判断该花是属于鸢尾花的哪一个品种。
短短十来行代码，已经到达了90%多的准确率，可以发现sklearn的代码效率是非常高的。

在实际的使用过程中，没有这么方便的数据可以直接调用，数据往往需要经过大量的处理。数据的处理往往占据了百分之七八十的工作量，在这一步中，你需要有扎实的python和panddas基础，这个步骤没有捷径，需要自己多写代码来熟悉了。
