### 一、 组合管理理论

#### 1. 必要收益率 = 实际无风险收益率 + 预期通货膨胀溢价 + 风险溢价

$k = r_f  + \pi^e + r_p$
$r_f ：无风险收益率$
$\pi^e ：预期通货膨胀率$
$r_p ：风险溢价$

#### 2. 现代投资组合理论（MPT）

对于两种资产组合：

$E(RP)=w_{1} E(R_{1} )+w_{2} E(R_{2} )$
$\sigma _{p} =\sqrt{w _{1}^{2} \sigma _{1}^{2}+w _{2}^{2} \sigma _{2}^{2}+2w _{1} w _{2} \sigma _{1}\sigma _{2}\rho _{1,2} }$
相关系数对投资组合效用的影响：

有效边界：不断增加投资组合中的资产数量，得到最左上方的一条曲线；有效边界上的投资组合为有效组合，包含了所有风资产，有效组合是完全分散化了的投资组合。

投资者的效用函数：
效用价值：$U = E(r)- 0.005A\sigma^2$

将二者结合，无差异曲线和有效边界的切点就是投资者效用最大的投资组合，为最优组合

```python
# 最小波动率
import scipy.optimize as sci_op

def fun_volatility_min(weights):
    covariance_matrix = rets.cov()
    expected_volatility = np.dot(np.dot(weights.T, covariance_matrix), weights)*252
    return np.sqrt(expected_volatility)

constraints = {'type':'eq', 'fun':lambda x: np.sum(x)-1.}
bounds = tuple((0,1) for x in range(5))
weights = np.random.random(5)
weights /= np.sum(weights)
initial_guess = weights

opts = sci_op.minimize(fun=fun_volatility_min,
                       x0=initial_guess,
                       method='SLSQP',
                       bounds=bounds,
                       constraints=constraints)
print('最小波动率', opts['fun'])
print (opts)
```

```python
# 给定方差求目标收益率最大
target = 0.35 #波动率为0.35

def fun_volatility_min(weights):
    covariance_matrix = rets.cov()
    expected_volatility = np.dot(np.dot(weights.T, covariance_matrix), weights)*252
    return np.sqrt(expected_volatility)

def fun_return_max(weights):
    expected_return = np.dot(weights.T, rets.mean())*252
    return -expected_return

constraints = ({'type':'eq', 'fun':lambda x: np.sum(x)-1.},
               {'type':'eq', 'fun':lambda x: fun_volatility_min(x)-target})

bounds = tuple((0,1) for x in range(5))
weights = np.random.random(5)
weights /= np.sum(weights)
initial_guess = weights

opts = sci_op.minimize(fun=fun_return_max,
                       x0=initial_guess,
                       method='SLSQP',
                       bounds=bounds,
                       constraints=constraints)
print('波动率上限为:',target,'最大预期收益率为' ,-1*opts['fun'])
print (opts)
```

### 3. 资本市场理论（CMT）

William Sharpe在MPT基础上引入无风险资产得到

如果我们想要计算出无风险资产与任一有效边界上的投资组合X组合后的情况：

$E(R_{p} )=w_{x} E(R_{x} )+(1-w_{x}) E(R_{f} )=R_{f}+w_{x}[E(R_{x}) -R_{f}]（1）$

$\sigma _{p}^{2} =w _{x}^{2} \sigma _{x}^{2}+(1-w _{x})^{2} \sigma _{f}^{2}+2w _{x}(1- w _{x}) \sigma _{x}\sigma _{f}\rho _{x,f}$

$E(R_{p} )=R_{f}+ \sigma _{p}\frac{ [E(R_{x}) -R_{f}]}{ \sigma _{x}}$
截距为R_{f}，斜率为$\frac{ [E(R_{x}) -R_{f}]}{ \sigma _{f}}$，斜率就是投资组合X的 Sharpe ratio

当这条直线与有效边界相切时，切点为M，这条直线将优于与任意有效组合X组合后得到的直线，也优于有效边界。FM取代了马科维茨有效边界，成为了新的有效边界。

此时，直线为：$E(R_{p} )=R_{f}+ \sigma _{p}\frac{ [E(R_{m}) -R_{f}]}{ \sigma _{m}}$

这条直线就是资本市场线（capital market line，CML）

点M为市场组合（market portfolio）

有无风险资产和风险资产组合形成资本配置线(CAL)
基本思路为投资者首先确定一个高风险资产的有效组合，然后再组合中加入无风险资产。
    


$CAL--E(r_c)=xE(r_p)+(1-x)r_f=r_f + x[E(r_p)-r_f]$
$E(r_c)$z组合预期收益率
$E(r_p)$风险资产组合收益率
$r_f$无风险资产收益率
$x$投资比例


不同的风险资产与无风险资产的配置，会形成不同的资本配置线。在均衡的情况下，投资者会选择最陡峭（有限制条件的）一条资本配置线CAL，
这条线被称为资本市场线CML，与有效边界的切点即为市场组合M。
$CML-- E(r_p) =r_f + \frac{E(r_M) - r_{f}}{\sigma_M}\sigma_p $
$r_M, \sigma_M$表示市场，可以设置基准，也可以数值模拟出来
如：
$f(x) = r_f + f'(x)\cdot x$
$f(0) = r_f$

### CAPM模型假设
- 1 投资者通过预期收益和方差来描述和评价资产或者资产组合，并按照马科维茨均值方差模型确定其单一期间的有效投资组合；对所有投资者而言投资起始期间都相同。
- 2 投资者为理性的个体，服从不满足和风险厌恶假设。
- 3 存在无风险利率，投资者可以按该利率进行借贷，并且对所有投资者而言，无风险利率都是相同的。
- 4 不存在任何手续费、佣金，也没有所得税及资本利得。即市场不存在任何交易成本。
- 5 所有投资者都能同时自由迅速地得到有关信息，即资本市场是有效率的。
- 6 所有投资者关于证券的期望收益率、方差和协方差、经济局势都有一致的预期。这也是符合马科维茨魔性的。依据马科维茨模型，给定一系列证券的价格和无风险利率，所有投资者对证券的预期收益率和协方差矩阵都相等，从而产生唯一的有效边界和独一无二的最优资产组合。这一假设也称为“同质期望”假设。

### $\beta系数$
$E(r_i) - r_f = \beta_i[E(r_M) - r_f]$
$\beta_i = cov(r_i, r_M)/  \sigma_M^2$

### 某证券风险
$\sigma_A^2 = \beta_A^2 \sigma_M^2 + \sigma_{\varepsilon A}^2$

### SML证券市场线, 使用$\alpha系数度量资产价格和期望收益率的不均衡状态。$
$SML: E(r_i) = r_f + [E(r_M) - r_f] \cdot \beta_i$

### CAPM在公司财务中的应用
$p = \frac{预期售价}{1+r_f+\beta(r_M - r_f)} $

### 零贝塔模型
使用均值-方差边界线非有效边界部分的投资组合替换无风险利率组成新的投资组合
$E(r_i) = E[r_{Z(M)}] + E[r_M - r_{Z(M)}]\frac{cov(r_i, r_M)}{\sigma_M^2}   $

### 流动性CAPM
$r_L$高流动性证券，流动费用$c_L$,净预期收益率为$E(r_l) - c_l/h$ 
$r_I$低流动性证券
非流动性溢价$r_I - r_L = \frac{1}{h_{LI}}(c_I - c_L) $
流动性效应的CAPM：$E(r_i) - r_f = \beta_i [E(r_M) - r_f] + f(c_i)  $

# 因素模型
<br />

#### 因素模型 是指一种假设证券的回报率只与不同的因素波动(相对数)或者指标的运动有关的证券定价模型，是一种回报率产生过程。
- 因素中的因子应该是系统影响所有证券价格的经济因素
- 在构造因素模型中，我们假设两个证券的回报率相关，一起变动，仅仅是因为它们对于因子变动的共同反应导致的。
- 证券回报率中不能由因子模型解释的部分是该证券所独有的，从而与别的证券回报率的特有部分无关，也与因素的变动无关。

### 一、单因素模型
$r_i = E(r_i) + m_i + e_i$
$E(r_i)$是证券持有期期初的预期收益  
$m_i$ 是证券持有期内非预期的宏观风险对证券的证券收益的影响
$e_i$ 是非预期的公司特有事情的影响。
<br />

不同的企业对于宏观经济因素的变化具有不同的敏感性。
记宏观因素的非预期成分为$F$，证券$i$对该因素的敏感度为$\beta_i$则
$r_i = E(r_i) + \beta_iF + e_i$
上式为单因素模型(single - factor model)
该证券(或组合)收益率的方差为
$\sigma_i^2 = \beta_i^2\sigma_F^2 + \sigma_{ei}^2 $
其中。$\beta_i^2$为因素风险，$\sigma_{ei}^2$为非因素风险。
<br />

- 单因素模型能够简化均值-方差计算量
- 单因素模型具有风险分散：
一个投资组合的风险因素即包含了系统风险：$\beta_p = \sum_{i=1}^N \omega_i\beta_i$,   
又包含了非系统性风险：$\sigma_{ep}^2 = \sum_{i=1}^N \omega_i^2 \sigma_{ei}^2 $   
组合内证券的数量增加，系统风险将平均化，非系统风险（都是相互独立的）将越来越小。   
<br />

 



### 二、单指数模型
单因素模型没有给出因素F的具体测度和明确界定。  
由夏普提出的单指数模型认为，可以用市场指数作为一般宏观因素的有效代表，   
影响资产价格波动的主要因素是市场总体价格水平(价格指数)。
<br / >

$r_i - r_f = \alpha_i + \beta_i(r_M - r_f) + e_i$
$\alpha_i,它是指当市场超额收益r_M - r_f 为0时证券i的期望超额收益率$
$\beta_i(r_M - r_f)，即证券i收益受到整个市场因素影响的部分$   
$e_i, 证券发行公司的特有因素对证券i的收益影响。$
<br / >

$R$代表超额收益   
$R_i = \alpha_i + \beta_i R_M + e_i$
$\sigma_i^2 = \beta_i^2 \sigma_M^2 + \sigma^2(e_i)$
组合中任意两种证券i和j的相关系数则可以表达为
$cov(r_i, r_j) = cov(\alpha_i + \beta_i R_M + e_i, \alpha_j + \beta_j R_M + e_j) $   
其中$\alpha_i, \alpha_j$是常数，因此它们与任何变量的相关系数为0   
又由于公司特有因素(e_i, e_j)与市场无关，且相互之间无关。   
因此两种证券之间的相关系数仅仅源于它们共同的依赖的因素$R_M$   
$cov(r_i, r_j) = cov(\beta_i R_M, \beta_j RM) = \beta_i \beta_j \sigma_M^2$

### 三、 多因素模型
单因素模型的一个隐含假定是，每个证券对每个风险因素具有相同的敏感度，但实际上不同的证券对不同的宏观因素具有不同的$\beta$值。   
当风险对期望收益有影响时，这一风险即是“可定价”的。   
多因素模型一般形式为：
$r_i = \alpha_i + \beta_{i1}F_{1t} +  \beta_{i2}F_{2t} + \beta_{i3}F_{3t} + ......+ e_{it}$

# 套利定价理论
在一个均衡资本市场中，所以的资产将遵从“一价法则”，即同一个资产即便在不同的市场上也只有一个均衡价格。   
当“一价法则”被违反的时候，即出现“套利机会”，   
<br \>
### 套利类型
<br \>
##### 无风险套利
- 空间套利
- 时间套利
- 工具套利：利用同一标的资产的现货以及各种衍生证券的价格差异，通过低买高卖来赚取无风险利润。在这种套利行为中，多种资产或者金融工具组合在一起，形成一种或多种与原来截然不同性质的复合金融工具。

##### 风险套利 
- 利用风险定价上的差异，通过低买高卖赚取无风险利润的交易行为。   

##### 税收套利
- 不同的投资主体、不同的证券、不同的收入来源以在税收待遇上存在的差异所进行的套利交易。

### 无套利均衡
套利机会大量和持续的出现，意味着市场处于非均衡的状态，此时投资者的套利行为将最终消除套利的机会，使市场恢复均衡。   
这一状态下的市场称为无套利均衡，它是指即使很少的投资者能够发现套利机会，并动用大笔资金获利，也能通过价格变动，很快恢复均衡的结果。
<br \>

### 套利投资组合的条件
<br \>
- 应是一个不需要投资者增加额外自己的组合。以$\Delta_i$表示投资者对证券$i$的持有量的改变，   
$\Delta_1 + \Delta_2 + ...... + \Delta_n = 0$

<br \>

- 该组合对任何因素都没有敏感性，即组合不存在额外风险。   
$\beta_{pj}=0$ , 这正是无风险套利的原因。

<br \>

- 组合的预期收益率为正  
$x_1E(r_1) + x_2E(r_2) + ...... + x_nE(r_n) > 0$

套利是利用同一种失误资产或证券的不同价格来赚取无风险利润的行为。
如果投资者可以找到这样一种证券组合，其初始净投资为零而且用能赚的正收益，   
那么所有的投资者都会去投资与这种吸引人的证券，结果这种证券组合的价格将会发生变化，   
直到均衡状态下整的收益降为零。
当这种交易不再存在时，就失去了套利机会，并且得到一种与资本资产定价模型类似的风险-收益模型。

### 套利定价理论的假设和主要观点
###### 一、假设
- 市场是完全竞争的、无摩擦的。 ———— 保证套利的可实施性
- 投资者是非满足的。追求效用最大化的：当投资者具有套利机会时，它们会构造套利证券组合来增加自己的财富。 ————进行套利的主观性
- 所有投资者有相同的预期，特别是，投资者都一致认为任意证券的收益率都是影响该证券收益率的k个因素的线性函数。 ————预期不同属风险套利
- 市场上的证券的种类远远大于因子的数目k。 ————保证有足够多的证券来分散掉不同的风险。
- 随机误差项为非系统风险，与所有影响因素正交。

##### 二、主要观点
- 套利定价是利用同一资产的不同价格赚取无风险利润的行为； 在一个高度竞争的、流动性强的市场中，套利行为将导致差价的消失，最终是市场趋于均衡。
- APT理论认为，套利行为是市场效率(市场均衡)的决定因素之一。如果市场未达到均衡，市场上即存在套利机会，投资者即会利用差价买入或者卖出，直至套利机会消失，市场恢复或达到均衡。
- 套利机会主要表现于差价的存在，因此凡是影响价格的因素都会影响套利的机会是否存在。
- 套利机会主要表现于差价的存在，因此凡是影响价格的因素都会影响套利机会是否存在。
- 根据无套利均衡原则，在因素模型下，具有相同因素敏感性的资产(组合)应提供相同的期望收益率。

### 套利模型的核心是对于未来收益率的精准预测：
假设一个充分分散化的投资组合A，其$\beta_A = 1$， 预期收益率为10%，则该投资组合的收益为：
$E(r_A) + \beta_A F$ = 10% + $1F $
在假设存在另一个投资组合B，其预期收益率为8%， $\beta_B = 1$   
如果组合A和组合B同时存在，将导致套利机会的出现。
做多资产组合A，做空B，可以获得2%的无风险套利。
$[(0.1 + 1F)Cash] - [(0.08 + 1F)Cash] = 0.02Cash $
