　核心竞争力是企业的立身之本。形成、强化、持续发展核心竞争力，应该是任何一个企业孜孜以求的。具有很强竞争力的公司往往能够持续的为社会创造价值，如苹果和谷歌。这类公司往往是值得投资的。本篇内容将企业的竞争优势量化，构造了一个衡量企业竞争力的因子--竞争优势因子。从逻辑出发以构建竞争优势因子为例讲解了构建具有经济学意义的因子的方法，深入浅出地讲解了因子的计算方法，并对因子有效性做了系统地分析，也为因子研究提供了一个模板。

# **1、前言：2017年因子风格纵览**

    2017年小盘股弱于大盘股，蓝筹股走强，因子风格出现转变，具体表现为：
        （1）估值，盈利和成长类等基本面Aalpha因子表现抢眼。
        （2）市场资金集中于大盘周期和行业白马龙头导致市值因子和反转因子失效。
    因子风格的转变使很多机构的Alpha策略失效，很多主动量化类基金对风格切换的把握不够及时，导致量化产品整体表现不佳。

    对于市值因子来说，小市值股票的红利逐渐消失，但因其市值较小，成长性较高，所以还具有成长优势。而大盘股经过一年大涨，估值优势已不再明显，今年出现的大市值股票走强的极端情况也难以长期持续。我们认为市值的alpha属性会慢慢向成熟市场的风险属性转变，更多呈现震荡特征。
    
    Alpha因子里面，估值盈利和成长这三类基本面因子属于“国际通行”的因子，在香港、美国和国内市场这些因子都很有效，而且各个市场的IC相差不大，这些因子是alpha的基本来源，而且更多体现的是投资者承担市场风险而获得的收益补偿，而并非市场错误定价而导致的短期套利收益，这是可以长期稳定存在的。

    基本面Alpha因子中的盈利类因子主要有，ROE，ROA等在财报中挖掘的因子，基于基本面因子走强，我们参考兴业证券的《基本面Alpha的复兴，上市公司竞争优势研究》，聚焦于上市公司的财务数据，通过对公司竞争优势维度的定量刻画，来把握基本面信息，构建一个选股因子-竞争优势因子。

# **2、竞争优势因子的构建**

## **2.1、公司竞争力的描述**

	竞争优势理论，是由哈佛大学商学研究院迈克尔·波特提出。波特认为，企业的竞争优势表现为企业的经营在产业竞争中高人一筹，也就是相同行业中，一些企业可以通过较低的成本获得较高的收益即在一定时期内实现高于行业平均水平的盈利。企业的竞争力有可能来自于规模优势、品牌优势、创新优势、渠道优势等。
    
    无论是哪一种优势，最终的体现总是公司的盈利能力提升，或者说相同行业，可以互相对比的公司，单位净资产能取得更多的利润。而衡量公司的竞争优势，往往是通过对比体现出来的。
    
    因此，我们根据以上对公司竞争力的描述，从一个方面来构建了公司竞争优势因子。通过对同一行业的公司在一个时间横截面的盈利排名和上一期年度公司的盈利排名来观察公司的竞争优势是否提高。
    
## **2.2、上市公司竞争优势因子的构建**
    
    根据以上对公司竞争优势的讨论，对上市公司竞争优势因子的构建提供了灵感。
    下面我们就来根据研报探讨并详细构建上市公司的竞争优势因子：
    
    
    第一步：计算ROE的TTM值。我们采用ROE的TTM值来作为判断公司过去一年的经营状况的标准，在这里直接用到优矿提供的因子库数据，API如下：（以平安银行为例,提取2017年10月16日到2017年10月30日的ROE数据，可以看到2017年10月23日平安银行的ROE发生变化，是因为三季报在当天发布。）

```python
DataAPI.MktStockFactorsDateRangeProGet(secID=u"",ticker=u"000001",beginDate=u"20171015",endDate=u"20171030",field=u"ticker,secID,tradeDate,ROE",pandas="1")
```

	第二步：计算行业内的ROE排名。考虑到各行业内上市公司数目不一，我们用公司在所属行业内ROE的排名除以该行业公司数目作为当期的ROE排名（相当于对排名做了一个归一化）。其中，为了使ROE排名越高，因子值越大。我们对公司的ROE从小到大排列，即ROE越小的公司，排名越小。ROE最小的公司，排名是1。
    （以银行业为例，计算2017年10月20日的ROE排名）

```python
all_ticker_sw = DataAPI.EquIndustryGet(industryVersionCD=u"010303",field=u"ticker,secShortName,industryVersionCD,isNew,industryName1",pandas="1")
#取出所有股票的申万行业分类
all_ticker_new = all_ticker_sw.loc[all_ticker_sw['isNew']==1]
#取出股票的最新分类
bank_ticker_sw = all_ticker_new.loc[all_ticker_new['industryName1']=='银行']
bank_ticker_sw.reset_index(drop=True,inplace=True)
bank_ticker_sw
```

```python
import pandas as pd
all_roe = DataAPI.MktStockFactorsOneDayGet(tradeDate=u"20171020",secID=u"",ticker=u"",field=u"ticker,tradeDate,ROE",pandas="1")
#取出所有股票在2017年10月20日的ROE
bank_roe = pd.merge(bank_ticker_sw,all_roe,on='ticker',how='inner')
#取出银行业在2017年10月20日的ROE
bank_roe.drop(['industryVersionCD','isNew'],axis=1,inplace=True)
#去掉没有用的信息
bank_roe_rank = bank_roe.sort_values(by='ROE').reset_index(drop=True).reset_index()
bank_roe_rank['RANK']=bank_roe_rank['index']+1
bank_roe_rank.drop('index',axis=1,inplace=True)
bank_num = len(bank_roe_rank)
bank_roe_rank['RANK_t'] = bank_roe_rank['RANK']/bank_num
#按照ROE计算排名
bank_roe_rank
```

	第三步、计算ROE的TTM值排名的变化。
    由于上市公司的业绩通常存在季节效应。为了排除季节效应影响，我们计算ROE排名的同比变化。研报中采用的是用当期的因子排名减去上年同期的因子排名。
    
    注意：
    根据研报的描述，在当期报表全部披露后才能计算ROE的当期正确排名，如果采用这种处理方法，在最后一家公司报表披露之前，因子值都是空缺的。我采用的方法是，无论报表是否为新披露的，都对当日的ROE进行排序(即在报表披露期间的ROE有新有旧）。当某一行业所有公司的报表全部披露后，这种方式计算的因子值也是正确的，在报表披露之前，因子值也不为空。在一定程度上也能反映因子排名的变化。
    第三步中计算同期相减应该是在所有报表披露后，再做同期相减。考虑到即使相同公司，在不同年份披露报表的披露时间也不同，计算起来非常复杂，为简单起见，直接用当日的ROE_t减去前一年的相同日期的ROE_t来得到排名变化。若前一年的该日期不是交易日，则用前一年该日期的后一个交易日的值来代替。
    下面，将计算ROE排名变化的程序封装成函数

```python
def cal_rankt(tradeDate=u'20170509'):
    all_ticker_sw = DataAPI.EquIndustryGet(industryVersionCD=u"010303",field=u"ticker,secShortName,industryVersionCD,isNew,industryName1",pandas="1")
#取出所有股票的申万行业分类
    all_ticker_new = all_ticker_sw.loc[all_ticker_sw['isNew']==1]
#取出股票的最新分类
    all_ticker_new.drop(['isNew','industryVersionCD'],axis=1,inplace=True)
    #找到最新的分类信息
    all_roe = DataAPI.MktStockFactorsOneDayGet(tradeDate=tradeDate,secID=u"",ticker=u"",field=u"ticker,tradeDate,ROE",pandas="1")
    #取出所有股票的某一天的ROE因子值
    roe_ind = pd.merge(all_ticker_new,all_roe,on='ticker',how='inner')
    rank_all = roe_ind.groupby('industryName1').apply(ind_rank)
    rank_all.reset_index(drop=True,inplace=True)
    return rank_all

def ind_rank(data):
    data = data.sort_values(by='ROE').reset_index(drop=True).reset_index()
    data['RANK'] = data['index']+1
    data.drop('index',axis=1,inplace=True)
    ind_num = len(data)
    data['RANK_t'] = data['RANK']/ind_num
    return data
```

```python
rankt_2017 = cal_rankt(u'20170428')#2017排名变化计算
rankt_2016 = cal_rankt(u'20160428')#2016排名变化计算
rankt_2016.rename(columns={'ROE':'ROE_LY','RANK':'RANK_LY', 'RANK_t':'RANK_t_LY','tradeDate':'tradeDate_LY'},inplace=True)
factor_df = pd.merge(rankt_2017,rankt_2016,on=['ticker','secShortName','industryName1'])
factor_df.loc[:,'FACTOR'] = factor_df['RANK_t']-factor_df['RANK_t_LY']
bank_factor = factor_df.loc[factor_df['industryName1']=='银行'].reset_index()
bank_factor
```

	以上我们以银行为例详细讲述了如何构建竞争优势因子，接下来给出计算2007年1月1日-2017年12月31日的竞争优势因子的代码。

```python
from pandas import DataFrame
import pandas as pd
import copy

start_date = '20070101'
end_date = '20171231'
cal_dates = DataAPI.TradeCalGet(exchangeCD=u"XSHG", beginDate=start_date, endDate=end_date).sort('calendarDate')
cal_dates = cal_dates.loc[cal_dates['isOpen']==1][['calendarDate','isMonthEnd']].reset_index(drop=True) #取出开盘交易日期
dates = cal_dates['calendarDate'].apply(lambda x:x.replace('-','')).tolist()
roe_data = []
for tradeDate in dates:
    tempt = DataAPI.MktStockFactorsOneDayGet(tradeDate=tradeDate,field=u"ticker,tradeDate,ROE",pandas="1")
    roe_data.append(tempt)
roe_df = pd.concat(roe_data,ignore_index=True)
all_ticker_sw = DataAPI.EquIndustryGet(industryVersionCD=u"010303",field=u"ticker,isNew,industryID1",pandas="1")
all_ticker_sw = all_ticker_sw.loc[all_ticker_sw['isNew']==1][['ticker','industryID1']]
roe_ind = pd.merge(all_ticker_sw,roe_df,on='ticker',how='inner')
```

```python
#对缺失ROE进行处理，用上一期的ROE来填补本期缺失ROE值
def pre_data(data):
    data = data.sort_values(by='tradeDate')
    data = data.fillna(method='pad')
    return data
roe_ind = roe_ind.groupby('ticker').apply(pre_data)
roe_ind.reset_index(drop=True, inplace=True)
roe_ind.dropna(inplace=True)
```

```python
#每天都要对每个行业的股票按ROE排序
def daily_rank(data):
    data = data.sort_values(by='ROE').reset_index(drop=True).reset_index()
    data['RANK'] = data['index']+1
    data.drop('index', axis=1, inplace=True)
    ind_num = len(data)
    data['RANK_t'] = data['RANK']/ind_num
    return data
roe_ind = roe_ind.groupby(['tradeDate','industryID1']).apply(daily_rank)
roe_ind.reset_index(drop=True,inplace=True)
roe_rank = roe_ind[['tradeDate','ticker','RANK_t']]
```

```python
#与上年同期相减
roe_rank_ly = copy.deepcopy(roe_rank)
roe_rank_ly['tradeDate'] = roe_rank_ly['tradeDate'].apply(lambda x:str(int(int(x[0:4])+1))+x[4:])
roe_rank_ly.rename(columns={'RANK_t':'RANK_t_LY'}, inplace=True)
rank_cg = pd.merge(roe_rank, roe_rank_ly, on=['tradeDate','ticker'],how='left')
#对上一年排名和当年排名未对齐情况进行处理
def pre_data2(data):
    data = data.sort_values(by='tradeDate')
    data = data.fillna(method='bfill')
    data = data.fillna(method='pad')
rank_cg = rank_cg.groupby('ticker').apply(pre_data2)
rank_cg = rank_cg.reset_index(drop=True)
rank_cg['value'] = rank_cg['RANK_t'] - rank_cg['RANK_t_LY']
rank_cg.drop(['RANK_t','RANK_t_LY'], axis=1, inplace=True)
rank_cg.rename(columns={'tradeDate':'date'},inplace=True)
```

```python
#构建月度因子
month_cal = cal_dates.loc[cal_dates['isMonthEnd']==1].reset_index(drop=True)
month_cal.rename(columns={'calendarDate':'date'},inplace=True)
month_factor = pd.merge(rank_cg, month_cal,on='date',how='inner')
month_factor['date'] = month_factor['date'].apply(lambda x:int(x.repalce('-','')))
month_factor['ticker'] = month_factor['ticker'].apply(lambda x:int(x))
month_factor.to_csv('month_factor.csv')
```

    下面将因子分为五组，分别查看五组因子的历史平均值。

```python
n_quantile = 5
cols_mean = ['meanF'+str(i+1) for i in range(n_quantile)]
cols = cols_mean
factor_means = pd.DataFrame(index=factor_df.index, columns=cols)
for dt in factor_means.index:
    qt_mean_results=[]
    tempt = factor_df.ix[dt].dropna()
    
    pct_quantiles = 1.0/n_quantile
    for i in range(n_quantile):
        down = tempt.quantile(pct_quantiles*i)
        up = tempt.quantile(pct_quantiles*(i+1))
        mean_tmp = tempt[(tempt<=up)&(tempt>=down)].mean()
        qt_mean_results.append(mean_tmp)
    factor_means.ix[dt] = qt_mean_results
fig = plt.figure(figsize=(16, 6))
ax1 = fig.add_subplot(111)

lns1 = ax1.plot(factor_means.index, factor_means.meanF1, label='F1')
lns2 = ax1.plot(factor_means.index, factor_means.meanF2, label='F2')
lns3 = ax1.plot(factor_means.index, factor_means.meanF3, label='F3')
lns4 = ax1.plot(factor_means.index, factor_means.meanF4, label='F4')
lns5 = ax1.plot(factor_means.index, factor_means.meanF5, label='F5')

lns = lns1+lns2+lns3+lns4+lns5
labs = [l.get_label() for l in lns]
ax1.legend(lns, labs, bbox_to_anchor=[0.5, 0.1], loc='', ncol=3, mode="", borderaxespad=0., fontsize=12)
ax1.set_ylabel(u'因子值', fontproperties=font, fontsize=16)
ax1.set_xlabel(u'日期', fontproperties=font, fontsize=16)
ax1.set_title(u"竞争优势因子历史表现", fontproperties=font, fontsize=16)
ax1.grid()
```

## **3.3 因子的预测指标--信息系数**

    为了评价因子是否具有预测能力，我们需要看看t时刻的因子值和t+1时刻的股票收益是否具有显著的相关性。信息系数IC正是这样一个指标，IC的定义是某时点因子在全部股票的因子暴露值与其下期回报的截面相关系数。
    IC = corr(factor(t),ret(t+1))
    按照上面的公式，只有从历史数据中才可以计算出信息系数IC。因为这要用到未来数据，即在t时刻，需要知道t时刻的股票因子值，和未来的t+1时刻相应股票的收益率。
    这个概念是站在现在的角度来回顾历史，看看该因子在历史上是否有效，但要注意我们用它来作为评价标准，是建立在未来因子的表现和过去是相同的，这样一个假设之上的，因为我们始终相信历史总是惊人的相似。但在一些特殊情况(如市场风格切换)下，这个假设便不再成立。
    为了计算信息系数，我们需要先计算因子在未来的收益率。在这里我计算的是往前一个月的收益率，即计算t时刻的因子暴露值和一个月后的股票收益率的相关系数。

```python
def getFowardReturns(universe, begin, end, window):
    end2 = datetime.datetime.strptime(end,'%Y%m%d')
    end2 = end2+datetime.timedelta(days=2*window) #为了保证计算出（begin到end）中最后一天的收益率，需要多取出一些数据，这里我设定多取出2倍未来区间长度的数据
    end2 = end2.strftime('%Y%m%d')
    #计算universe中所有股票，在日期在begin和end之间的（包含起止日期），未来window天的收益率。
    data = DataAPI.MktEqudAdjGet(secID=universe, beginDate=begin, endDate=end, isOpen='1',field=['ticker', 'tradeDate', 'closePrice'])
    def WinRetCal(mktdf,win):
        mktdf = mktdf.sort('tradeDate')
        mktdf['forwardReturns'] = mktdf['closePrice'].shift(-win)/mktdf['closePrice']-1.0
        return mktdf[['tradeDate','ticker', 'forwardReturns']]
    secWinRet = data.groupby('ticker').apply(lambda x:WinRetCal(x,window)).reset_index(drop=True)
    be = begin[0:4]+'-'+begin[4:6]+'-'+begin[6:]
    en = end[0:4]+'-'+end[4:6]+'-'+end[6:]
    return secWinRet.loc[(secWinRet['tradeDate']>=be)&(secWinRet['tradeDate']<=en)]

#计算全A股票的未来20个交易日(近乎一个月）的股票收益率
universe = set_universe('A')
secReturn = getFowardReturns(universe, begin='20080101', end='20171031',window=20)
factor_df = factor_df = pd.read_csv('month_factor.csv')[['date','ticker','value']]
factor_df['date']=factor_df['date'].apply(str)
factor_df['date']=factor_df['date'].apply(lambda x:x[0:4]+'-'+x[4:6]+'-'+x[6:])
factor_df['ticker'] = factor_df['ticker'].map(lambda x:str('%06d'%x))
factor_df = factor_df.rename(columns={'date':'tradeDate'})
faReDf = pd.merge(factor_df, secReturn, on=['ticker','tradeDate'],how='inner')
# 计算个股因子值和未来收益的相关系数
def cal_corr(df):
    factorValue = df['value']
    fowardReturn = df['forwardReturns']
    return factorValue.corr(fowardReturn)

all_ic = faReDf.groupby('tradeDate').apply(cal_corr)
all_ic.index = pd.to_datetime(all_ic.index)
fig = plt.figure(figsize=(18,7))
ax = fig.add_subplot(111)
all_ic.plot()
ax.set_title(u"竞争优势因子的信息系数IC", fontproperties=font, fontsize=18)
ax.grid()
```

    从结果中可以看到，在2014年以后因子的IC值大多数在大于零，说明因子具有正向的预测能力。

## **3.4 因子的分组超额收益**

	我们把因子从大到小分为五组，观察每组因子的平均收益，通过这种方法可以判断因子对股票的区分能力。

```python

mean_re = faReDf['forwardReturns'].mean()
def mean_recal(datadf):
    datadf = datadf.dropna()
    n_quantile = 5
    cols_mean = [i+1 for i in range(n_quantile)]
    cols = cols_mean
    pct_quantile = 1.0/n_quantile
    re_mean = []
    for i in range(n_quantile):
        down = datadf['value'].quantile(pct_quantile*i)
        up = datadf['value'].quantile(pct_quantile*(i+1))
        mean_tmp = datadf.loc[(datadf['value']<=up)&(datadf['value']>=down),'forwardReturns'].mean()-mean_re
        re_mean.append(mean_tmp)
    return DataFrame([re_mean],columns=cols)

excess_returns_means = faReDf.groupby('tradeDate').apply(mean_recal).reset_index().drop('level_1',axis=1).set_index('tradeDate')
fig = plt.figure(figsize=(12, 6))
ax1 = fig.add_subplot(111)

excess_returns_means_dist = excess_returns_means.mean()
excess_dist_plus = excess_returns_means_dist[excess_returns_means_dist>0]
excess_dist_minus = excess_returns_means_dist[excess_returns_means_dist<0]
lns2 = ax1.bar(excess_dist_plus.index, excess_dist_plus.values, align='center', color='r', width=0.35)
lns3 = ax1.bar(excess_dist_minus.index, excess_dist_minus.values, align='center', color='g', width=0.35)

ax1.set_xlim(left=0.5, right=len(excess_returns_means_dist)+0.5)
ax1.set_ylim(-0.01, 0.004)
ax1.set_ylabel(u'超额收益', fontproperties=font, fontsize=16)
ax1.set_xlabel(u'五分位分组', fontproperties=font, fontsize=16)
ax1.set_xticks(excess_returns_means_dist.index)
ax1.set_xticklabels([int(x) for x in ax1.get_xticks()], fontproperties=font, fontsize=14)
ax1.set_yticklabels([str(x*100)+'0%' for x in ax1.get_yticks()], fontproperties=font, fontsize=14)
ax1.set_title(u"竞争优势因子超额收益", fontproperties=font, fontsize=16)
ax1.grid()
```

	从这里看到，三分位和四分位的平均收益为正，其他都是负的，这说明因子选股能力一般，对股票没有很好的区分能力。

# **4、 因子改进方案**
	
    前面我们根据上市公司在其所属行业内ROE排名的变化构建了上市公司竞争优势因子，并且对因子的预测能力和选股能力进行了分析。在分析过程中，我们发现，虽然因子有一定的选股能力，但因子仍有待改进。在这里我参考研报，提出改进方案，改进后的因子分析就不做了。
## **4.1 上市公司竞争优势的分解**	
	
    (1)我们可以利用杜邦分解将公司的ROE分解到净利润率，资产周转率，和权益乘数三个值的乘积。所以，ROE的提升可能来自于净利润率提升、资产周转率的提升和权益乘数的提高。
    
    (2)净利润率又称销售净利率，是反映公司盈利能力的一项重要指标。它体现公司的定价权，以及销售费用。净利润率的提升反映了公司定价权的提升，或者公司对费用管理能力的提高。而定价权的提升在竞争优势中表现为品牌形象的提升，技术先进等。
    
    (3)资产周转率的提升表现为资产的使用和管理效率的提升，包括固定资产使用效率，存货管理和其他营运资本的管理的提升等，反映的是企业相对于同行业其他公司在资产使用效率方面的优势的提升。
    
    (4)权益乘数的计算公式是总资产除以所有者权益，它反映了公司的杠杆率。公司的杠杆率越高，则公司的权益乘数越大。而杠杆不能反映公司的竞争优势。
    
    (5)通过对ROE的三个因子分别分析讨论发现，净利润率和资产周转率都能反映公司的竞争优势，而权益乘数则不能。若剔除权益乘数的影响，竞争优势因子能够更有效地反映公司的竞争优势。

### **ROE（净利润/所有者权益） = 净利润（净利润/营业收入） × 资产周转率（营业收入/总资产） × 权益乘数（总资产/所有者权益）**

## 4.2 **竞争优势因子的改进**

	基于以上的讨论，我们仅从净利润率和资产周转率两个方面来构建新的竞争优势因子作为对上述因子的改进，而且单季度更能反映公司近期优势的变化，对近期股价影响较大，所以改进后我们采用单季度的财务数据进行排名计算：
    
    （1）计算单季度净利润率排名。按照上面计算ROE排名的方式计算行业内净利润率的排名。
    
    （2）计算单季度资产周转率排名。同样按照上面计算ROE排名的方式计算行业内净资产周转率的排名。
    
    （3）计算综合排名。将净利润率排名和净资产周转率的排名等权相加，得到综合排名。
    
    （4）计算综合排名的变化。计算综合排名同比的变化作为改进之后的竞争优势因子值。
    
    上面的构建过程同时考虑了公司的净利润率和资产周转率的影响，排除了公司的财务杠杆的影响，能更加清晰的反映公司的竞争能力。
