# A股市场情况

```python
import numpy as np # 时间序列的计算
import pandas as pd #  处理结构化的表格数据
import datetime as dt  
from pandas import Series, DataFrame, isnull      
from datetime import timedelta, datetime
import matplotlib.pyplot as plt
import seaborn as sns  #用于美化matplotlib图表
sns.set_style('whitegrid') #改变背景颜色
from CAL.PyCAL import *


secID = set_universe('A')
beginDate = '20170101'
endDate = '20170407'
cal = Calendar('CHINA.SSE')
time1 = cal.bizDatesList(beginDate, endDate)
field=['tradeDate','ticker','secShortName','chgPct','closePrice','turnoverRate','turnoverVol','turnoverValue','negMarketValue','marketValue','isOpen']
jg = DataFrame()
for tradetime in time1:
    data = DataAPI.MktEqudGet(tradeDate=tradetime,secID=secID,ticker=u'',beginDate='',endDate='',isOpen=u"",field=field,pandas="1")
    result = {'tradetime' : tradetime,  #交易时间
              'tingpai' : data[data['isOpen']==0]['tradeDate'].count(),  #停牌家数
              'zhuangting' : data[data['chgPct']>0.099]['tradeDate'].count(),  #涨停家数
              'dieting' : data[data['chgPct']<-0.099]['tradeDate'].count(),    #跌停家数
              'total_turnoverRate' : data['turnoverRate'].sum()*100,          #总换手率
              'total_turnoverVol' : data['turnoverVol'].sum()/1e8,            #总成交量
              'total_turnoverValue' : data['turnoverValue'].sum()/1e8,        #总成交额
              'total_negMarketValue' : data['negMarketValue'].sum()/1e12,       #总流通市值
              'total_MarketValue' : data['marketValue'].sum()/1e12,             #总市值
              'pos_money_ratio' : float(data[data['chgPct']>0]['tradeDate'].count())/float(data[data['isOpen']==1]['tradeDate'].count())*100,  #赚钱效应
              'neg_money_ratio' : float(data[data['chgPct']<0]['tradeDate'].count())/float(data[data['isOpen']==1]['tradeDate'].count())*100,   #亏钱效应
              'neg_chgPct_top50' : data.sort(['chgPct'], ascending=True).head(50)['chgPct'].mean()*100,   #前50平均跌幅
              'pos_chgPct_top300' : data[data['chgPct']<0.11].sort(['chgPct'], ascending=False).head(300)['chgPct'].mean()*100,     #前300平均涨幅
              'total_turnoverValue/MarketValue' : data['turnoverValue'].sum()/data['marketValue'].sum(),       #总成交金额与市值比
              'turnoverRate_top50' : data.sort(['turnoverRate'], ascending=False).head(50)['chgPct'].mean()*100      #换手前50平均涨幅
              
             }
    
    jg = jg.append(DataFrame(result, index=[tradetime]))
    
```
