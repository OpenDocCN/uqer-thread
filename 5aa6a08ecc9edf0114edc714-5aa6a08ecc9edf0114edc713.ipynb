```python
import time
import json
import pandas as pd
import numpy as np
import datetime as dt
import scipy.stats as st
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('white')
from CAL.PyCAL import *

import itertools
import copy

DataAPI.settings.cache_enabled = True
```

```python
import pandas as pd
pool = set_universe('A')
income_A=pd.DataFrame(columns=["secID","publishDate","endDate","endDateRep","secShortName","tRevenue","NIncomeAttrP"])
income_S1=pd.DataFrame(columns=["secID","publishDate","endDate","endDateRep","secShortName","tRevenue","NIncomeAttrP"])
income_Q1=pd.DataFrame(columns=["secID","publishDate","endDate","endDateRep","secShortName","tRevenue","NIncomeAttrP"])
income_CQ3=pd.DataFrame(columns=["secID","publishDate","endDate","endDateRep","secShortName","tRevenue","NIncomeAttrP"])
for i in pool:
    # 先处理年报
    tmp_A=DataAPI.FdmtISGet(ticker=u"",secID=i,reportType="A",endDate=u"",beginDate=u"",
                      field=u"secID,publishDate,endDate,endDateRep,secShortName,tRevenue,NIncomeAttrP",pandas="1")
    tmp_A.drop_duplicates(subset='endDate',inplace=True)
    income_A=pd.concat([tmp_A,income_A], ignore_index=True)
    # 再处理半年报
    tmp_S1=DataAPI.FdmtISGet(ticker=u"",secID=i,reportType="S1",endDate=u"",beginDate=u"",
                      field=u"secID,publishDate,endDate,endDateRep,secShortName,tRevenue,NIncomeAttrP",pandas="1")
    tmp_S1.drop_duplicates(subset='endDate',inplace=True)
    income_S1=pd.concat([tmp_S1,income_S1], ignore_index=True)
    # 一季报
    tmp_Q1=DataAPI.FdmtISGet(ticker=u"",secID=i,reportType="Q1",endDate=u"",beginDate=u"",
                      field=u"secID,publishDate,endDate,endDateRep,secShortName,tRevenue,NIncomeAttrP",pandas="1")
    tmp_Q1.drop_duplicates(subset='endDate',inplace=True)
    income_Q1=pd.concat([tmp_Q1,income_Q1], ignore_index=True)
    # 三季报
    tmp_CQ3=DataAPI.FdmtISGet(ticker=u"",secID=i,reportType="CQ3",endDate=u"",beginDate=u"",
                      field=u"secID,publishDate,endDate,endDateRep,secShortName,tRevenue,NIncomeAttrP",pandas="1")
    tmp_CQ3.drop_duplicates(subset='endDate',inplace=True)
    income_CQ3=pd.concat([tmp_CQ3,income_CQ3], ignore_index=True)
```

```python
def get_code(time,case):     #time is report time, case is A or S1.
    hs300=set_universe('HS300',date=time)
    cyb=set_universe('CYB',date=time)
    zxb=set_universe('ZXB',date=time)
    if case == 'S1':   # 半年报
        CYB_S1=income_S1[income_S1['secID'].isin(cyb)]
        ZXB_S1=income_S1[income_S1['secID'].isin(zxb)]
        HS300_S1=income_S1[income_S1['secID'].isin(hs300)]
        return CYB_S1,ZXB_S1,HS300_S1
    if case == 'A':    # 年报
        CYB_A=income_A[income_A['secID'].isin(cyb)]
        ZXB_A=income_A[income_A['secID'].isin(zxb)]
        HS300_A=income_A[income_A['secID'].isin(hs300)]
        return CYB_A,ZXB_A,HS300_A
    if case == 'Q1':   # 一季报
        CYB_Q1=income_Q1[income_Q1['secID'].isin(cyb)]
        ZXB_Q1=income_Q1[income_Q1['secID'].isin(zxb)]
        HS300_Q1=income_Q1[income_Q1['secID'].isin(hs300)]
        return CYB_Q1,ZXB_Q1,HS300_Q1
    if case == 'CQ3':   # 三季报单季
        CYB_CQ3=income_CQ3[income_CQ3['secID'].isin(cyb)]
        ZXB_CQ3=income_CQ3[income_CQ3['secID'].isin(zxb)]
        HS300_CQ3=income_CQ3[income_CQ3['secID'].isin(hs300)]
        return CYB_CQ3,ZXB_CQ3,HS300_CQ3

time_S1=list(income_S1[income_S1.endDate>'2010-01-01'].endDate.drop_duplicates())
time_A=list(income_A[income_A.endDate>'2010-01-01'].endDate.drop_duplicates())
time_Q1=list(income_Q1[income_Q1.endDate>'2010-01-01'].endDate.drop_duplicates())
time_CQ3=list(income_CQ3[income_CQ3.endDate>'2010-01-01'].endDate.drop_duplicates())

df_S1=pd.DataFrame(columns=['time','hs300_NI','hs300_TR','hs300','cyb_NI','cyb_TR','cyb','zxb_NI','zxb_TR','zxb'])
df_A=pd.DataFrame(columns=['time','hs300_NI','hs300_TR','hs300','cyb_NI','cyb_TR','cyb','zxb_NI','zxb_TR','zxb'])
df_Q1=pd.DataFrame(columns=['time','hs300_NI','hs300_TR','hs300','cyb_NI','cyb_TR','cyb','zxb_NI','zxb_TR','zxb'])
df_CQ3=pd.DataFrame(columns=['time','hs300_NI','hs300_TR','hs300','cyb_NI','cyb_TR','cyb','zxb_NI','zxb_TR','zxb'])
tmp=0
# 处理半年报
for t in time_S1:
    CYB_S1,ZXB_S1,HS300_S1=get_code(t,'S1')
    df_S1.loc[tmp,'time']=t
    df_S1.loc[tmp,'hs300_NI']=HS300_S1[HS300_S1.endDate==t].NIncomeAttrP.sum()
    df_S1.loc[tmp,'hs300_TR']=HS300_S1[HS300_S1.endDate==t].tRevenue.sum()
    df_S1.loc[tmp,'cyb_NI']=CYB_S1[CYB_S1.endDate==t].NIncomeAttrP.sum()
    df_S1.loc[tmp,'cyb_TR']=CYB_S1[CYB_S1.endDate==t].tRevenue.sum()
    df_S1.loc[tmp,'zxb_NI']=ZXB_S1[ZXB_S1.endDate==t].NIncomeAttrP.sum()
    df_S1.loc[tmp,'zxb_TR']=ZXB_S1[ZXB_S1.endDate==t].tRevenue.sum()
    tmp+=1
    
    
tmp=0
# 处理年报
for t in time_A:
    CYB_A,ZXB_A,HS300_A=get_code(t,'A')
    df_A.loc[tmp,'time']=t
    df_A.loc[tmp,'hs300_NI']=HS300_A[HS300_A.endDate==t].NIncomeAttrP.sum()
    df_A.loc[tmp,'hs300_TR']=HS300_A[HS300_A.endDate==t].tRevenue.sum()
    df_A.loc[tmp,'cyb_NI']=CYB_A[CYB_A.endDate==t].NIncomeAttrP.sum()
    df_A.loc[tmp,'cyb_TR']=CYB_A[CYB_A.endDate==t].tRevenue.sum()
    df_A.loc[tmp,'zxb_NI']=ZXB_A[ZXB_A.endDate==t].NIncomeAttrP.sum()
    df_A.loc[tmp,'zxb_TR']=ZXB_A[ZXB_A.endDate==t].tRevenue.sum()
    tmp+=1
# 处理一季报
for t in time_Q1:
    CYB_Q1,ZXB_Q1,HS300_Q1=get_code(t,'Q1')
    df_Q1.loc[tmp,'time']=t
    df_Q1.loc[tmp,'hs300_NI']=HS300_Q1[HS300_Q1.endDate==t].NIncomeAttrP.sum()
    df_Q1.loc[tmp,'hs300_TR']=HS300_Q1[HS300_Q1.endDate==t].tRevenue.sum()
    df_Q1.loc[tmp,'cyb_NI']=CYB_Q1[CYB_Q1.endDate==t].NIncomeAttrP.sum()
    df_Q1.loc[tmp,'cyb_TR']=CYB_Q1[CYB_Q1.endDate==t].tRevenue.sum()
    df_Q1.loc[tmp,'zxb_NI']=ZXB_Q1[ZXB_Q1.endDate==t].NIncomeAttrP.sum()
    df_Q1.loc[tmp,'zxb_TR']=ZXB_Q1[ZXB_Q1.endDate==t].tRevenue.sum()
    tmp+=1
    
    
tmp=0
# 处理三季报
for t in time_CQ3:
    CYB_CQ3,ZXB_CQ3,HS300_CQ3=get_code(t,'CQ3')
    df_CQ3.loc[tmp,'time']=t
    df_CQ3.loc[tmp,'hs300_NI']=HS300_CQ3[HS300_CQ3.endDate==t].NIncomeAttrP.sum()
    df_CQ3.loc[tmp,'hs300_TR']=HS300_CQ3[HS300_CQ3.endDate==t].tRevenue.sum()
    df_CQ3.loc[tmp,'cyb_NI']=CYB_CQ3[CYB_CQ3.endDate==t].NIncomeAttrP.sum()
    df_CQ3.loc[tmp,'cyb_TR']=CYB_CQ3[CYB_CQ3.endDate==t].tRevenue.sum()
    df_CQ3.loc[tmp,'zxb_NI']=ZXB_CQ3[ZXB_CQ3.endDate==t].NIncomeAttrP.sum()
    df_CQ3.loc[tmp,'zxb_TR']=ZXB_CQ3[ZXB_CQ3.endDate==t].tRevenue.sum()
    tmp+=1
```

```python
# 得到指数相应的涨跌幅
time=list(income_S1.endDate.drop_duplicates())+list(income_A.endDate.drop_duplicates())+\
        list(income_Q1.endDate.drop_duplicates())+list(income_CQ3.endDate.drop_duplicates())
df_index=pd.DataFrame(columns=['cyb','hs300','zxb'],index=time)
df_index=df_index.sort_index()
df_index=df_index[df_index.index>'2010-04-01']
tmp=0
for t in list(df_index.index):
    df_index.ix[t,'cyb']=DataAPI.MktIdxdGet(endDate=t,ticker=u"399006",pandas="1").closeIndex.iloc[-1]
    df_index.ix[t,'hs300']=DataAPI.MktIdxdGet(endDate=t,ticker=u"000300",pandas="1").closeIndex.iloc[-1]
    df_index.ix[t,'zxb']=DataAPI.MktIdxdGet(endDate=t,ticker=u"399005",pandas="1").closeIndex.iloc[-1]
    
df_index=df_index.pct_change()
df_index=df_index[df_index.index>'2011-01-01']
df_index
```

```python
df_S1_yoy=df_S1.set_index('time')
df_S1_yoy=df_S1_yoy.sort_index()
df_S1_yoy=df_S1_yoy.pct_change()

df_A_yoy=df_A.set_index('time')
df_A_yoy=df_A_yoy.sort_index()
df_A_yoy=df_A_yoy.pct_change()

df_Q1_yoy=df_Q1.set_index('time')
df_Q1_yoy=df_Q1_yoy.sort_index()
df_Q1_yoy=df_Q1_yoy.pct_change()

df_CQ3_yoy=df_CQ3.set_index('time')
df_CQ3_yoy=df_CQ3_yoy.sort_index()
df_CQ3_yoy=df_CQ3_yoy.pct_change()

df=df_A_yoy.append(df_S1_yoy).append(df_Q1_yoy).append(df_CQ3_yoy)
df=df.sort_index()
df=df.drop('2017-12-31')   # 因为2017年的年报还没有出完。

df.hs300=df_index.hs300
df.cyb=df_index.cyb
df.zxb=df_index.zxb

df=df.dropna()
df
```
