```python
import pandas as pd
import numpy as np

#取交易日
from CAL.PyCAL import *
datedata=DataAPI.TradeCalGet(exchangeCD=u"XSHG",beginDate='20090112',endDate='20170301',field=['calendarDate','isOpen']) 
datedata=datedata[datedata['isOpen'] == 1]
date = map(lambda x: x[0:4]+x[5:7]+x[8:10], datedata['calendarDate'].values.tolist())    #多取一个月是为了判断是否在前二十个交易日上市

#取所有A股
stocks = DataAPI.EquGet(equTypeCD='A',listStatusCD='L',field='secID,nonrestfloatA')
universe = stocks['secID'].tolist()
```

```python
#取A股股票上市的数据
IPOinf=pd.DataFrame()      #用于存储处理后的股票ID和上市日期
IPO=DataAPI.EquIPOGet(eventProcessCD=u"",secID=universe,field=u"secID,listDate")
IPOinf[['secID','listDate']]=IPO[['secID','listDate']]             
IPO=IPO.set_index('secID')    #以股票ID为index                      
for i in range(len(universe)):
    IPOinf['listDate'][i]=int(filter(lambda x:x.isdigit(),IPO.ix[universe[i],0]))  #把上市日期处理成整型，方便以后调用处理
IPOinf=IPOinf[IPOinf['listDate']>=20090112]                      #取在我们研究时间范围内的上市数据
IPOinf=IPOinf.sort(columns='listDate').set_index('listDate')          #以上市日期为index，方便查找在该日期上市的股票ID
```

```python
#获取涨停板及涨停板第二日各价情况
uplimitinf=pd.DataFrame()   #用于存储所有涨停板及第二日数据
for i in range(len(date)-31):   #对日期进行循环
    
    delist=[]  #用于存储当日及前二十个交易日内上市的股票
    for j in range(30):
        if date[i+30-j] in IPOinf.index:
            delist=delist+IPOinf.ix[int(date[i+30-j]),0].values.tolist()
    universe=list(set(universe).difference(set(delist)))   #从全A股里剔除掉当日及前二十个交易日内上市的股票
    
    #判断当日涨停的股票，我这里使用的方法是判断当日收盘价是否等于涨停价
    data1=DataAPI.MktLimitGet(secID=universe,tradeDate=date[i+30],field=u"ticker,limitUpPrice,upLimitReachedTimes").set_index('ticker')
    data1=data1[data1['upLimitReachedTimes']>0]     #先剔除一部分当日没有涨停的股票
    data2=DataAPI.MktEqudGet(ticker=data1.index,tradeDate=date[i+30],field=u"ticker,closePrice").set_index('ticker')
    data3=pd.concat([data1,data2],axis=1)
    data3=data3[data3['closePrice'] == data3['limitUpPrice']]   #判断当日收盘价是否等于涨停价
    
    #调用当日涨停股票的前复权的股价信息，把第二日的各价信息存储到uplimitinf里，方便以后处理
    s=DataAPI.MktEqudAdjGet(tradeDate=date[i+31],ticker=data3.index,isOpen="1",field=u"ticker,tradeDate,preClosePrice,openPrice,highestPrice,lowestPrice,closePrice")
    uplimitinf=pd.concat([uplimitinf,s],axis=0)
```

```python
#先加载画图需要用的包
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.style.use('ggplot')
import seaborn as sns
uplimitinf = pd.read_csv("uplimitinformation.csv")
```

```python
#用increase存储各价涨幅情况
increase=pd.DataFrame()
increase['increaseOpen']=uplimitinf['openPrice']/uplimitinf['preClosePrice']-1
increase['increaseHigh']=uplimitinf['highestPrice']/uplimitinf['preClosePrice']-1
increase['increaseLow']=uplimitinf['lowestPrice']/uplimitinf['preClosePrice']-1
increase['increaseClose']=uplimitinf['closePrice']/uplimitinf['preClosePrice']-1
```

```python
plt.hist(np.array(increase['increaseOpen']),color=sns.desaturate("indianred", .8))
plt.title('increaseOpen')
```

```python
print 'Above 0:'+str(float(len(increase[increase['increaseOpen']>0]))/len(increase)), 'mean:'+str(increase['increaseOpen'].mean())
```
