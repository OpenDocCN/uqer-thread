```python
def risk_plot(df, figsize=(16,8)):
    
    df1 = df[[u'主动收益']].sort_values(by=u'主动收益')
    df2 = df[[u'主动权重']].sort_values(by=u'主动权重')
    df3 = df[[u'主动风险']].sort_values(by=u'主动风险')
    
    fig = plt.figure(figsize=figsize)
    ax = fig.add_subplot(131)
    fig.set_tight_layout(True)
    ax = df1.plot(kind='barh', color='yellowgreen', ax=ax)
    labels = [label for label in df1.index.values] 
    ax.set_yticklabels(labels, fontproperties=font, fontsize=13, rotation=0)
    ax.legend(prop=font, loc=4)
    ax.set_title(u'主动收益', fontproperties=font, fontsize=15)

    ax2 = fig.add_subplot(132)
    ax2 = df2.plot(kind='barh', color='orange', ax=ax2)
    labels = [label for label in df2.index.values] 
    ax2.set_yticklabels(labels, fontproperties=font, fontsize=13, rotation=0)
    ax2.legend(prop=font, loc=4)
    ax2.set_title(u'主动权重', fontproperties=font, fontsize=15)

    ax3 = fig.add_subplot(133)
    ax3 = df3.plot(kind='barh', color='dodgerblue', ax=ax3)
    labels = [label for label in df3.index.values] 
    ax3.set_yticklabels(labels, fontproperties=font, fontsize=13, rotation=0)
    ax3.legend(prop=font, loc=4)
    ax3.set_title(u'主动风险', fontproperties=font, fontsize=15)
```

```python
# 基于风险模型的多期行业归因
# accumulated: 单期归因，还是多期累计归因  True=多期累计，False=单期归因

fund_ia = fa.industry_attribution(fund_, start_, end_, benchmark=benchmark_, accumulated=True)
risk_plot(fund_ia)
```

```python
# 基于风险模型的多期风格归因(accumulated=True)
fund_sa_all = fa.style_attribution(fund_, start_, end_, benchmark=benchmark_, accumulated=True)
risk_plot(fund_sa_all, (16, 5))
```

```python
# 基于风险模型的单期风格归因(accumulated=False)
fund_sa_case = fa.style_attribution(fund_, start_, end_, benchmark=benchmark_, accumulated=False)
fund_sa_case = fund_sa_case.tail(13)
print 'date: ', fund_sa_case.index.get_level_values(0)[-1]
fund_sa_case.index = fund_sa_case.index.droplevel()

risk_plot(fund_sa_case, (16, 5))
```
