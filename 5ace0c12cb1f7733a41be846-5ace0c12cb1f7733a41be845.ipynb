# Pandas中DataFrame数据结构构建方法汇总

```python
#字典式构建方法，columns和数据
#字典套列表
from pandas import Series,DataFrame
import pandas as pd
DF={"name":["zhangshan","lisi","wangwu"],"age":[12,13,14],"sex":["male","female","male"]}
f1=DataFrame(DF)
print f1
```

```python
#字典式构建方法，columns和数据和索引
#字典套列表
from pandas import Series,DataFrame
import pandas as pd
DF={"name":["zhangshan","lisi","wangwu"],"age":[12,13,14],"sex":["male","female","male"]}
f1=DataFrame(DF,index=["one","two","three"])
print f1
```

```python
#字典式构建方法，columns和数据,同时确认column顺序
#字典套列表
from pandas import Series,DataFrame
import pandas as pd
DF={"name":["zhangshan","lisi","wangwu"],"age":[12,13,14],"sex":["male","female","male"]}
f1=DataFrame(DF,index=["one","two","three"],columns=["name","age","sex"])  #此处可添加更多的column
print f1
```
