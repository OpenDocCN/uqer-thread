# 计算HM模型
公式
$$
R_i-R_f=\alpha+\beta _1 (R_m-R_f)+D\beta_2(R_m-R_f)^2 +\epsilon\\\D=1:当R_m>R_f\\\D=0:当R_m<R_f
$$

```python
# 通联数据
DataAPI.FundPerfIndicGet(secID='159928.XSHE',dataDate='20210803',
                                  window=u"1",field=u"secID,selectionHM,timingHM",pandas="1")
```

```python
#取得基金净值
data=DataAPI.FundNavGet(secID=['159928.XSHE','510300.XSHG'],beginDate=u"20210701",
                         endDate=u"20210803",field=u"secID,endDate,ADJUST_NAV",pandas="1")
data=data.groupby(['secID','endDate']).mean().unstack().T
data=data.loc['ADJUST_NAV']
data=data/data.shift(1)-1#每日涨跌幅
data=data-0.025#R-Rf
data.dropna(inplace=True)
data['rm2']=pow((data['510300.XSHG']+abs(data['510300.XSHG']))/2,2)
data.columns=[['y','x1','x2']]
```

```python
import statsmodels.api as sm
Y = data['y']
X = data[['x1','x2']]
result = sm.OLS(Y, sm.add_constant(X)).fit()
```
