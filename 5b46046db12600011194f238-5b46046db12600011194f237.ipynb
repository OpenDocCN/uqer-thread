# 随机变量

使用choice生成100个随机数

```python
import numpy  as np
import pandas as pd
import matplotlib.pyplot as plt
from  scipy import stats

randNum=np.random.choice([1,2,3,4,5],size=100,replace=True,p=[0.1,0.1,0.2,0.3,0.3])
randNum
```

```python
#频数分布
pd.Series(randNum).value_counts()
```

```python
#频率分布
pd.Series(randNum).value_counts()/100
```

```python
begin='20160101'
end='20180101'

data=DataAPI.MktIdxdGet(indexID=u"",ticker=u"000300",tradeDate=u"",beginDate=begin,endDate=end,exchangeCD=u"XSHE,XSHG",field=u"tradeDate,ticker,closeIndex",pandas="1")
```

```python
#添加收益率序列
data['return']=data['closeIndex'].pct_change().dropna()
#去除缺失值
data=data.dropna()
data['sig']=0.0

data.loc[data['return']>0.0,['sig']]=1
data.head()
```

```python
#核密度估计中的高斯分布
desity=stats.kde.gaussian_kde(data.iloc[:,-1])
#设定分割区间
bins=np.arange(-5,5,0.02)

fig=plt.figure(figsize=(8,6))
#概率密度曲线
ax1=fig.add_subplot(211)
ax1.plot(bins,desity(bins))

#累计概率密度曲线
ax2=fig.add_subplot(212)
ax2.plot(bins,desity(bins).cumsum())
plt.show()
```
