# 基于隐马尔可夫模型的市场择时策略

```python
from hmmlearn.hmm import GaussianHMM
from hmmlearn.hmm import GMMHMM
import datetime 
import numpy as np
from matplotlib import cm, pyplot as plt
import matplotlib.dates as dates
import pandas as pd
import seaborn as sns
import random
sns.set_style('white')
```

```python
beginDate = '20100601'
endDate = '20170628'
data = DataAPI.MktEqudGet(ticker=u"300070",beginDate=beginDate,endDate=endDate,field=['tradeDate','closePrice','highestPrice','lowestPrice','turnoverVol'],pandas="1")
data1 = DataAPI.FstDetailGet(ticker=u"300070", beginDate=beginDate, endDate=endDate, field=['tradeVal'], pandas='1')
tradeVal = data1
tradeDate = pd.to_datetime(data['tradeDate'][5:])
volume = data['turnoverVol'][5:]
closeIndex = data['closePrice']
deltaIndex = np.log(np.array(data['highestPrice'])) - np.log(np.array(data['lowestPrice']))
deltaIndex = deltaIndex[5:]
logReturn1 = np.array(np.diff(np.log(closeIndex))) #收盘价的对数差，或称为对数收益率
logReturn1 = logReturn1[4:]
logReturn5 = np.log(np.array(closeIndex[5:])) - np.log(np.array(closeIndex[:-5])) #周对数收益率
logReturnFst = np.array(np.diff(np.log(tradeVal['tradeVal'])))[4:]
closeIndex = closeIndex[5:]
```

```python
trainx = np.column_stack([logReturn1[:1352], logReturn5[:1352], deltaIndex[:1352], volume[:1352], logReturnFst[:1352]])
testx = np.column_stack([logReturn1[1352:], logReturn5[1352:], deltaIndex[1352:], volume[1352:], logReturnFst[1352:]])
```

```python
random.seed(1)
model = GMMHMM(n_components=4, n_iter=1000).fit([trainx])
hidden_states = model.predict(trainx)
```

```python
print 'means and vars of each hidden state:'
for i in range(model.n_components):
    print('{0}th hidden state'.format(i))
    #print('mean = ',model.means_[i])
    #print('var = ', np.diag(model.covars_[i]))
```

```python
fig = plt.figure(figsize=(15,10))
for i in range(model.n_components):
    grh = 'ax' + str(i)
    grh = fig.add_subplot(model.n_components,1,i+1)
    idx = (hidden_states == i)
    grh.plot_date(tradeDate[idx], closeIndex[idx], '.', label='{0}th hidden state'.format(i), lw=1)
    plt.legend()
    plt.grid(1)
```

```python
res = pd.DataFrame({'tradeDate':tradeDate[:1393],'logReturn1':logReturn1[:1393],'logReturn5':logReturn5[:1393],'volume':volume[:1393],'state':hidden_states}).set_index('tradeDate')
#res
```

```python
fig2 = plt.figure(figsize = (15, 15))
for i in range(model.n_components):
    idx = (hidden_states == i)
    idx = np.append(0, idx[:-1])
    df = res.logReturn1
    res['sig_ret%s'%i] = df.multiply(idx, axis=0)
    gph = 'axs' + str(i)
    gph = fig2.add_subplot(6,1,i+1)
    gph.plot_date(tradeDate[:1393], np.exp(res['sig_ret%s'%i].cumsum()), '-', label='%dth hidden state'%i)
    plt.legend(loc='upper center')
    plt.grid(1)
```

```python
hidden_states2 = model.predict(testx)
res2 = pd.DataFrame({'tradeDate':tradeDate[1393:], 'logReturn1':logReturn1[1393:], 'logReturn5':logReturn5[1393:], 'volume':volume[1393:], 'state':hidden_states2}).set_index('tradeDate')
```

```python
plt.figure(figsize = (15, 8))
idxlong = (hidden_states2 == 2) + (hidden_states2 == 1) + (hidden_states2 == 4)
idxlong = np.append(0, idxlong[:-1])
idxshort = (hidden_states2 == 0) + (hidden_states2 == 3)
idxshort = np.append(0, idxshort[:-1])
df = res2.logReturn1
res2['sig_retbest2'] = df.multiply(idxlong, axis=0) - df.multiply(idxshort, axis=0)
plt.plot_date(tradeDate[1393:], np.exp(res2['sig_retbest2'].cumsum()), '-',label='strategy')
plt.plot_date(tradeDate[1393:], np.exp(res2['logReturn1'].cumsum()), '-',label='base')
plt.legend()
plt.grid(1)
```
