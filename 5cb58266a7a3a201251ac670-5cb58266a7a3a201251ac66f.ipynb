```python
# -*- coding:utf-8 -*-
from __future__ import division
import numpy as np
import pandas as pd
import matplotlib.pylab as plt
import seaborn as sns
import statsmodels.api as sm

from functools import wraps
def wrapper_func_cache(func):
    @wraps(func)
    def wrapped(*args, **kwargs):
        if not hasattr(func, '__cache_result'):
            func.__cache_result = {}
        
        hash_key = [str(i) for i in args] + [str(j) for j in kwargs.iteritems()]
        cache_key = hash(tuple(hash_key))
        
        if cache_key not in func.__cache_result:
            func.__cache_result[cache_key] = func(*args, **kwargs)
            
        return func.__cache_result[cache_key]
    return wrapped

@wrapper_func_cache
def Monthly_const_trade_calendar(beginDate=u"20050101",endDate=u""):
    """交易日历，可以选择cache，只调用 api 一次
    """
    data = DataAPI.TradeCalGet(exchangeCD=u"XSHG,XSHE", beginDate=beginDate, endDate=endDate, field=['calendarDate', 'isMonthEnd'])
    data = data[data.isMonthEnd == 1].calendarDate.tolist()
    data = sorted(list(set(data))) 
    return data

# ST过滤
def st_remove(source_universe, st_date=None):
    """
    给定股票列表,去除其中在某日被标为ST的股票
    Args:
        source_universe (list of str): 需要进行筛选的股票列表
        st_date (datetime): 进行筛选的日期,默认为调用当天
    Returns:
        list: 去掉ST股票之后的股票列表

    Examples:
        >> universe = set_universe('A')
        >> universe_without_st = st_remove(universe)
    """
    st_date = st_date if st_date is not None else datetime.datetime.now().strftime('%Y%m%d')
    df_ST = DataAPI.SecSTGet(secID=source_universe, beginDate=st_date, endDate=st_date, field=['secID'])
    return [s for s in source_universe if s not in list(df_ST['secID'])]

# bug 函数封装问题
def winsorize_median(tmp):
    if isinstance(tmp, pd.Series):
        mc = sm.stats.stattools.medcouple(tmp)
        data = sorted(tmp)
        q1 = np.percentile(data, 25)
        q3 = np.percentile(data, 75)
        iqr = q3-q1
        if mc >= 0:
            l = q1-1.5*np.exp(-3.5*mc)*iqr
            u = q3+1.5*np.exp(4*mc)*iqr
        else:
            l = q1-1.5*np.exp(-4*mc)*iqr
            u = q3+1.5*np.exp(3.5*mc)*iqr
        tmp[tmp<l]=l
        tmp[tmp>u]=u
    return tmp   

# bug 函数封装问题
def standardize_zscore(tmp):
    if isinstance(tmp, pd.Series):
        mu = tmp.mean()
        sigma = tmp.std()
        tmp = (tmp - mu)/sigma
    return tmp   

# 获取申万一级行业分类成份股编码 
def Stock_Industry(point):
    # 申万一级指数编码
    industrys=DataAPI.IndustryGet(industryVersion=u"SW",industryVersionCD=u"",industryLevel=u"1",isNew=u"1",field=u"",pandas="1")
    indexes=list(industrys['indexSymbol'].astype(str))#行业对应的指数编码
    index_param=DataAPI.IdxGet(secID=u"",ticker=indexes,field=u"ticker,secShortName,publishDate",pandas="1")#获取指数上市时间即为对应行业开始时间
    #2014年前已存在的行业
    old_industry=index_param[index_param['publishDate']<'2014-01-01']
    old_indexes=list(old_industry['ticker'])
    old_industry_name=list(old_industry['secShortName'])
    #2014年后的所有行业
    new_industry=index_param
    new_indexes=list(new_industry['ticker'])
    new_industry_name=list(new_industry['secShortName'])
    # 申万一级指数分类函数
    SW_Old = old_indexes
    SW_New = new_indexes    
    if isinstance(point,str):
        point = pd.to_datetime(point)
    timepoint = pd.to_datetime('2014-01-01')
    if point <= timepoint:
        return SW_Old
    elif point > timepoint:
        return SW_New
    else:
        print 'wrong'

# 获取申万一级行业分类成份股       
def Industry_Constituent_Stocks(industry, timepoint):
    tmp = DataAPI.IdxConsGet(ticker=industry, intoDate=timepoint, isNew=u"",field=['consID'],pandas="1")
    return sorted(list(tmp.consID.values))
```

## 因子 IC 值的计算 
因子的 IC 值是指因子在第 T 期的暴露度与 T+1 期的股票收益的相关系数，即 
$IC_d^T = corr(R^{T+1}, d^T)$
其中，$IC_d^T$代表因子d在第T期的IC值，$R^{T+1}$代表所有个股第T+1期的收益率向量，
$d^T$表所有个股第 T 期在因子 d 上的暴露度向量。
测试模型构建方法如下： 
1. 股票池、回溯区间、截面期均与回归法相同。 
2. 因子值在去极值、标准化、去空值处理后，在截面期上用其做因变量对市值因子及行业   
因子（哑变量）做线性回归，取残差作为因子值的一个替代。这样做可以消除行业因素    
和市值因素对因子的影响。 计算残差序列和 T+1 期股票收益序列的相关系数作为 T 期因   
子 IC 值。 
3. 因子评价方法：    
a) IC 值序列的均值大小——因子显著性；    
b) IC 值序列的标准差——因子稳定性；     
c) IR 比率（IC 值序列均值与标准差的比值）——因子有效性；     
d) IC 值累积曲线——随时间变化效果是否稳定；     
e) IC 值序列大于零的占比——因子作用方向是否稳定。     

```python
# 测试因子
fac = 'volatility'

IC = {}

for date in trade_calendar[:-1]:
    #print date
    # 获取下一次回测的日期
    date_next = trade_calendar[trade_calendar.index(date)+1]
    
    # 获取当日股票池
    univ = list(Ddata.loc[date].secID)
    univ = st_remove(univ, date)

    # 获取横截面股票收益率
    date_next = trade_calendar[trade_calendar.index(date)+1]
    R_T  = DataAPI.MktEqudGet(tradeDate=date, secID=univ, beginDate=u"",endDate=u"",isOpen="1",field=['secID','closePrice','preClosePrice','negMarketValue'],pandas="1")
    R_T ['return'] = (R_T.closePrice-R_T.preClosePrice) / R_T.preClosePrice
    R_T['fac_MV'] = standardize_zscore(winsorize_median(R_T.negMarketValue))
    R_T= pd.merge(R_T.loc[:,['secID','fac_MV','return']], Ddata.loc[date,['secID','volatility']], on='secID')
    
    nextmonthyreturn  = DataAPI.MktEqudGet(tradeDate=date_next, secID=univ, beginDate=u"",endDate=u"",isOpen="1",field=['secID','closePrice','preClosePrice'],pandas="1")
    nextmonthyreturn['nreturn'] = np.log(nextmonthyreturn.closePrice /nextmonthyreturn.preClosePrice)

    # 获取横截面因子载荷
    factor_loadings = DataAPI.MktStockFactorsOneDayGet(tradeDate=date, secID=univ, field=['secID'],pandas="1")
    factor_loadings[fac] = standardize_zscore(winsorize_median(R_T['volatility']))
    factor_loadings.set_index('secID',inplace=True, drop=False)
    #print factor_loadings

    # 获取行业暴露度 哑变量矩阵
    SW_Industry_List = Stock_Industry(date)
    Linear_Regression = pd.DataFrame()
    for i in SW_Industry_List:
        i_Constituent_Stocks = Industry_Constituent_Stocks(i, date)
        i_Constituent_Stocks = list(set(i_Constituent_Stocks).intersection(set(univ)))
        try:
            tmp = factor_loadings.loc[i_Constituent_Stocks]
            tmp.dropna(inplace=True)
            tmp[i]=1.0
        except:
            print i
        Linear_Regression = Linear_Regression.append(tmp)
    Linear_Regression.fillna(0.0, inplace=True)

    Linear_Regression = pd.merge(Linear_Regression, R_T.loc[:,['secID','return','fac_MV']], on='secID')
    Linear_Regression = pd.merge(Linear_Regression, nextmonthyreturn.loc[:,['secID','nreturn']], on='secID')
    Linear_Regression.set_index('secID', inplace=True)

    X = Linear_Regression.iloc[:,:28]
    X['fac_MV'] = Linear_Regression.fac_MV
    y = Linear_Regression.loc[:,fac]
    
    # OLS回归
    est=sm.OLS(y,X)      #残差项
    result = est.fit()
    
    # 因子 IC 值的计算 
    Inf_coeff =Linear_Regression  #.set_index('secID')
    Inf_coeff['Residual'] = result.resid
    m = Inf_coeff.loc[:,['Residual','nreturn']].corr().iloc[0,1]
    IC[date] = m

IC = pd.Series(IC)
IC.dropna(axis=0, how='any', inplace=True)


print 'IC 值序列的均值大小',IC.mean(),'IC 值序列绝对值的均值大小',np.mean(np.abs(IC))
print 'IC 值序列的标准差',IC.std()
print 'IR 比率（IC 值序列均值与标准差的比值）',IC.mean()/IC.std()
n = [x for x in IC.values if x>0]
print 'IC 值序列大于零的占比',len(n)/len(IC)

tmp = (IC+1.0).cumprod()-1.0
tmp.plot(figsize=(16,5))
```

## 分层模型回测 
- 分层模型主要是考量保序作用。    


依照因子值对股票进行打分，构建投资组合回测，是最直观的衡量指标优劣的手段。一般来
说，通过回归法和计算因子 IC 值都无法确定因子的单调性（例如，某因子值排名在中间 1/3
的个股表现比前 1/3 和后 1/3 的个股表现要好） ， 但是分层回测法是可以确定因子单调性的。
分层回测法逻辑简单，结果清晰，操作方便，并且具有能区分因子单调性的独特优势，是接
受度非常高的一种单因子测试手段。

```python
#############################################    行业中性化的五分位测试   ##############################################

fac = 'volatility'

# 可编辑部分与 strategy 模式一样，其余部分按本例代码编写即可

# -----------回测参数部分开始，可编辑------------
start = BEGIN                     # 回测起始时间
end = END                       # 回测结束时间
benchmark = 'HS300'                        # 策略参考标准
universe = DynamicUniverse('HS300')           # 证券池，支持股票和基金
capital_base = 10000000                     # 起始资金
freq = 'd'                              
refresh_rate = Monthly(1)  
accounts = {
    'fantasy_account': AccountConfig(account_type='security', capital_base=10000000)
}

# ------------------------------------------------ 将因子分组并回测 ---------------------------------------------------------

# 调用quick_backtest方法
# https://uqer.datayes.com/help/api/#quick_backtest参数优化
sim_params = quartz.SimulationParameters(start, end, benchmark, universe, capital_base, refresh_rate=refresh_rate, accounts=accounts)
# 获取回测行情数据
data = quartz.get_backtest_data(sim_params)
# 运行结果
results = {}

for quantile_five in range(0, 5):
    
    # ---------------策略逻辑部分----------------
    refresh_rate = Monthly(1)  
    commission = Commission(0.0002, 0.0002)
    
    def initialize(context):                   # 初始化虚拟账户状态
        pass
    
    def handle_data(context):               
        Operation_Date = context.current_date.strftime('%Y%m%d')
        account = context.get_account('fantasy_account')
        current_universe = context.get_universe('stock', exclude_halt=True)
        pre_date = context.previous_date.strftime("%Y-%m-%d")
       
        # 获取当日股票池
        univ = list(Ddata.loc[pre_date].secID)
        univ = st_remove(univ, pre_date)
        
        # 判断行业分类
        Industrys = Stock_Industry(context.current_date)

        # 获取行业划分成份股
        Constituent_Stocks = {}
        for s in Industrys:
            tmp = Industry_Constituent_Stocks(s, Operation_Date)
            tmp = list(set(tmp).intersection(set(univ)))
            Constituent_Stocks[s] = tmp

        # 获取因子数据
        factor_ = DataAPI.MktEqudGet(tradeDate=pre_date, secID=univ, beginDate=u"",endDate=u"",isOpen="1",field=['secID','closePrice','preClosePrice','negMarketValue'],pandas="1")
        factor_ = pd.merge(factor_.loc[:,['secID']], Ddata.loc[pre_date,['secID','volatility']], on='secID')
        factor_.dropna(inplace=True)
        factor_.set_index('secID', inplace=True)

       # 按行业排序筛选
        holding_list = []
        holding = {}
        for s in Industrys:
          #  i_Constituent_Stocks = Industry_Constituent_Stocks(s,Operation_Date)
          #  i_Constituent_Stocks = list(set(i_Constituent_Stocks).intersection(set(univ)))
            try:
                tmp = Constituent_Stocks[s]
                tmp = factor_.loc[tmp,:]
                tmp.sort_values(fac, inplace=True)
                q_min = tmp.quantile(quantile_five*0.2)[fac]
                q_max = tmp.quantile((quantile_five+1)*0.2)[fac]
                industry_univ = tmp[(tmp[fac]>=q_min)&(tmp[fac]<=q_max)]
                industry_univ = industry_univ.index.values.tolist()
                holding[s] = industry_univ
                holding_list = holding_list + industry_univ
            except:
                pass

        # 伪 行业中性
        neu_weights = {}
        neu = set_universe('HS300',context.previous_date.strftime('%Y%m%d')) + set_universe('ZZ500',context.previous_date.strftime('%Y%m%d'))
        tmp = DataAPI.MktEqudGet(tradeDate=context.previous_date.strftime('%Y%m%d'),secID=neu, isOpen="",field=['secID','negMarketValue'],pandas="1")
        tmp.set_index('secID', inplace=True)
        for s in Industrys:
            try:
                cm = list(set(neu).intersection(set(holding[s])))
                cm = tmp.loc[cm]
                cm = cm.negMarketValue.sum()
                neu_weights[s] = cm
            except:
                pass
        neu_sum = pd.Series(neu_weights).sum()

        # 仓位操作
        # 不在股票池中的，清仓
        positions = account.get_positions()
        sell_list = [stk for stk in positions if stk not in holding_list]
        for stk in sell_list:
            order_to(stk, 0)
        # 在目标股票池中的，等权买入
        for s in Industrys:
            try:
                neu_weight = neu_weights[s]/neu_sum
                order_univ = holding[s]
                for i in order_univ:
                    order_pct_to(i, 1.0/len(order_univ)*neu_weight)
            except:
                pass
                
# ---------------策略逻辑部分结束----------------
   # 把回测逻辑封装到 TradingStrategy 中，供 quick_backtest 使用
    strategy = quartz.TradingStrategy(initialize, handle_data)
    # 回测部分
    bt, perf = quartz.quick_backtest(sim_params, strategy, data=data)

    # 对于回测的结果，可以通过 perf_parse 函数计算风险指标
    perf = quartz.perf_parse(bt, perf)

    # 保存运行结果
    tmp = {}
    tmp['bt'] = bt
    tmp['annualized_return'] = perf['annualized_return']
    tmp['volatility'] = perf['volatility']
    tmp['max_drawdown'] = perf['max_drawdown']
    tmp['alpha'] = perf['alpha']
    tmp['beta'] = perf['beta']
    tmp['sharpe'] = perf['sharpe']
    tmp['information_ratio'] = perf['information_ratio']
    
    results[quantile_five] = tmp
    print str(quantile_five)
print 'done'   
```
