# scipy.optimize.minimize

scipy.optimize.minimize(fun, x0, args=(), method=None, jac=None, hess=None, hessp=None, bounds=None, constraints=(), tol=None, callback=None, options=None)


fun: 求最小值的目标函数

x0:变量的初始猜测值，如果有多个变量，需要给每个变量一个初始猜测值。minimize是局部最优的解法，所以，变量的初始猜测值可能会影响最终结果。

method:求极值的方法，官方文档给了很多种。一般使用默认。每种方法我理解是计算误差，反向传播的方式不同而已，这块有很大理论研究空间

constraints:约束条件，针对fun中为参数的部分进行约束限制

返回：

返回优化结果对象，x：优化问题的目标数组。success: True表示成功与否，不成功会给出失败信息。



```python
# 导入所需要的模块
# scipy.optimize.minimize是我们需要的优化器
from scipy.optimize import minimize
import numpy as np 
import matplotlib.pyplot as plt

# 下面两行让matplotlib可以显示中文
plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus']=False


#案例1
#计算使得函数f(x)= 1/x + x 取得最小值的x 
def fun(a):  
    v=lambda x:a/x[0] + x[0]     
    return v  
# a赋值为1
a=1
# 得到函数f(x)= 1/x + x 
fun=fun(a) 

# 猜测初始值
x0 = 2
# 使用minimize优化器计算
res = minimize(fun, x0, method='SLSQP')
print(res.fun)
print(res.success)
print(res.x)

"""res的结果及其解读如下：

     fun: 2.0000000815356342
     jac: array([ 0.00057095])
 message: 'Optimization terminated successfully.'
    nfev: 19
     nit: 6
    njev: 6
  status: 0
 success: True
       x: array([ 1.00028559])
       
目标函数的最小值为fun: 2.0000000815356342
目标参数的雅克比值（梯度向量）为：jac: array([ 0.00057095])
minimize优化器终止的原因是优化已经成功，message: 'Optimization terminated successfully.'
优化器对目标函数进行了19次优化，nfev: 19
优化器进行了两轮迭代，nit: 6
优化器对雅克比值（梯度向量）进行了两次优化，njev: 6
优化器终止时的状态值是0，status: 0，我认为，0代表优化器完成了优化任务
优化器是否成功完成了任务：是。success: True
优化器返回的最优参数是 x: array([ 1.00028559])
"""
```

```python
# 案例2
# 计算  函数f(x)=(2+x1)/(1+x2) - 3*x1+4*x3 的最小值  其中x1,x2,x3的范围在0.1到0.9 之间

def fun(a,b,c,d):    
    # 原本的函数表达式为v=lambda x: (a+x1)/(b+x2) -c*x1+d*x3   
    # 由于minimize优化器返回的最优参数(res.x，注意，这个地方的x不能用其他字符替代)是一个数组，所以，x1的位置用x[0]替代，x2的位置用x[1]替代，x3的位置用x[2]替代，
    v=lambda x: (a+x[0])/(b+x[1]) -c*x[0]+d*x[2]   
    return v
def con(args):    
    # 约束条件分为eq 和ineq    
    # eq表示函数结果等于0 ； ineq表 表达式大于等于0
    # args是一个元组或者列表，其中6个元素，按顺序分别对应X1最小和最大值，X2最小和最大值，X3最小和最大值，args将在后续函数运行中被赋值
    x1min, x1max, x2min, x2max,x3min,x3max = args    
    cons = (# 第一个约束条件{'type': 'ineq', 'fun': lambda x: x1 - x1min}的意思是，lambda x大于等于0，也就是x1 - x1min >= 0,即x1 >= x1min。x1min将在后续函数运行中赋值为0.1，所以，x1>=0.1。
        # 由于minimize优化器返回的最优参数(res.x，注意，这个地方的x不能用其他字符替代)是一个数组，所以，x1的位置用x[0]替代
        {'type': 'ineq', 'fun': lambda x: x[0] - x1min},
        # 第二个约束条件{'type': 'ineq', 'fun': lambda x: -x1 + x1max}的意思是，lambda x大于等于0，也就是-x1 + x1max >= 0,即x1 <= x1max。x1max将在后续函数运行中赋值为0.9，所以，x1<=0.9
        # 由于minimize优化器返回的最优参数(res.x，注意，这个地方的x不能用其他字符替代)是一个数组，所以，x1的位置用x[0]替代
        {'type': 'ineq', 'fun': lambda x: -x[0] + x1max},
        {'type': 'ineq', 'fun': lambda x: x[1] - x2min},
        {'type': 'ineq', 'fun': lambda x: -x[1] + x2max},
        {'type': 'ineq', 'fun': lambda x: x[2] - x3min},
        {'type': 'ineq', 'fun': lambda x: -x[2] + x3max})    
    return cons #cons就是我们的约束条件

#对函数的系数赋值    
(a,b,c,d)= (2,1,3,4)   
# 得到函数f(x)
fun=fun(a,b,c,d)

#设置参数范围/约束条件    
args = (0.1,0.9,0.1, 0.9,0.1,0.9)  #x1min, x1max, x2min, x2max    
# 得到约束函数
cons = con(args)    

#设置初始猜测值
x_start = np.array([0.5,0.5,0.5])
# 运行minimize优化器
res = minimize(fun, x_start, method='SLSQP',constraints=cons)
print(res.fun)
print(res.success)
print(res.x)
print(res)


"""# 案例2运行minimize的返回值
     fun: -0.773684210526435
     jac: array([-2.47368421, -0.80332409,  4.        ])
 message: 'Optimization terminated successfully.'
    nfev: 10
     nit: 2
    njev: 2
  status: 0
 success: True
       x: array([ 0.9,  0.9,  0.1])

# 解读如下
目标函数的最优值为fun: -0.773684210526435
目标参数的雅克比值（梯度向量）为：jac: array([-2.47368421, -0.80332409,  4.        ])
minimize优化器终止的原因是优化已经成功，message: 'Optimization terminated successfully.'
优化器对目标函数进行了10次优化，nfev: 10
优化器进行了两轮迭代，nit: 2
优化器对雅克比值（梯度向量）进行了两次优化，njev: 2
优化器终止时的状态值是0，status: 0，我认为，0代表优化器完成了优化任务
优化器是否成功完成了任务：是。success: True
优化器返回的最优参数是x: array([ 0.9,  0.9,  0.1])

"""
```

```python
# 案例3： 资产组合的有效边界求解
# 例如我们有四种资产，利用这四种资产构建投资组合，根据马克维茨的现代投资组合理论可以求解有效边界。求解有效边界的过程实际上就是一个求解最优化问题的过程。

# 四种资产的年化收益率分别为
r = np.array([0.15, 0.12, 0.31, 0.04])
# 四种资产的协方差矩阵为
cov_mat = np.array([[0.2148,0.147,0.0868,0.127],
                   [0.147,0.2063,0.1095,0.1176],
                   [0.0868,0.1095,0.2983,0.1015],
                   [0.127,0.1176,0.1015,0.1449]])
# 把年化收益率序列的len()值赋值给k
k = len(r)


# 为了更好突显有效边界，这里随机构造4000个投资组合
# 初始化投资组合回报、投资组合波动率，两者都是空列表
port_r, port_var = [], []
# 进行4000次迭代
for p in range(4000):
    # 生成随机的权重，权重元素的个数为k，由上面可知，这对应四种资产的年化收益率
    weights = np.random.random(k)
    # 把权重总和调整为1
    weights /= np.sum(weights)
    # 算出加权重之后投资组合收益率，并把该收益率append到投资组合回报列表中
    port_r.append(np.sum(r*weights))
    # 算出加权重之后投资组合方差，并把该方差append到投资组合方差列表中
    port_var.append(np.sqrt((weights.T).dot(cov_mat).dot(weights)))
    # 这里有一个问题，为什么weights.T要使用转置呢？由下面的实验可以看出，这里是否转置都行。可能是因为weights只是一维的数组，如果是二维形状的，就不行了。
    """
weights
[ 0.10144257  0.10700769  0.12937653  0.66217321]
weights.T
[ 0.10144257  0.10700769  0.12937653  0.66217321]
(weights.T).dot(cov_mat).dot(weights)
0.130109592285
(weights).dot(cov_mat).dot(weights)
0.130109592285
(weights).dot(cov_mat).dot(weights.T)
0.130109592285
len(weights)
4
weights.shape
(4,)"""

# 把投资组合回报、方差转换为数组，并分别赋值给port_r, port_var
port_r, port_var = np.array(port_r), np.array(port_var)
```

利用马克维茨的MPT——在组合目标收益率情况下，计算最小化组合方差和标准差，求解有效边界。具体而言就是在不同的期望回报率约束下，求解下面的数学优化问题


$$Min 　投资组合方差=\sum_{i=1}^{n}\sum_{j=1}^{n}w_iw_j\sigma_{ij}$$    
$$s.t. \sum_{i=1}^{n}w_iE(r_i) = 期望回报率, \sum_{i=1}^{n}w_i = 1$$

```python
# 自定义一个函数，可以得到投资组合回报、方差
def statistics(weights):
    # 把权重转化为数组
    weights = np.array(weights)
    # 计算投资组合回报
    port_returns = np.sum(r*weights)
    # 计算投资组合的方差
    port_variance = (weights.T).dot(cov_mat).dot(weights)
    # 函数返回投资组合的回报、方差
    return np.array([port_returns, port_variance])

# 自定义另一个函数，调用上面的函数，并返回投资组合的方差
def min_variance(weights):
    return statistics(weights)[1]

# 目标回报是一个列表，有50个元素，linspace函数把从0到0.5的数切出来50个，左闭右开
target_returns = np.linspace(0.06,0.3,50)
# 目标方差是一个空列表
target_variance = []

# 迭代目标回报列表中的每一个元素
for tar in target_returns:
    # bnds为限制条件，((0, 1), (0, 1), (0, 1), (0, 1))，weights中的单个元素w需满足0<w<1的条件
    bnds = tuple((0,1) for i in range(k))   
    # cons为约束条件；这里使用了lambda函数来表示优化问题的约束条件；x是权重向量，也就是statistics函数中的weights
    cons = (# eq表示相等，即lambda函数等于0；lambda x等于0，也就是statistics(x)[0]-tar等于0，即statistics(x)[0]=tar，就是statistics函数中的port_returns=tar
        {'type':'eq','fun':lambda x:statistics(x)[0]-tar},
        # eq表示相等，即lambda函数等于0；np.sum(x)-1，也就是所有权重的加和=1
        {'type':'eq','fun':lambda x:np.sum(x)-1})  
    # 启用minimize优化器求解
    res = minimize(min_variance, #目标函数，这里是自定义的最小方差函数min_variance
                   k*[1./k],#设置初始猜测值
                   method = 'SLSQP', #优化器使用的方法
                   bounds = bnds, #限制条件，权重中每个元素的值必须在（0，1）区间内
                   constraints = cons#约束条件，见上面的解读
                  )
    # 计算出目标收益率对应的最小方差值，并把这个数字添加到target_variance列表中
    target_variance.append(res['fun'])
# 把得到的target_variance列表转化为数组
target_variance = np.array(target_variance)
```

```python
# 画出随机投资组合和有效边界

# 通过方差计算得出标准差
standard_deviation=np.sqrt(target_variance)

risk_free = 0
plt.figure(figsize = (10,5))
plt.scatter(port_var, port_r, 
            c=(port_r-risk_free)/port_var,  marker = 'o',
            label=u'random portfolio',cmap='rainbow')
plt.plot(standard_deviation, target_returns, 
         c = 'red', linewidth=2.3, 
         label=u'efficient border' )
plt.grid(True)
plt.xlabel(u'volatility')
plt.ylabel(u'expected return')
plt.legend(loc='best')
plt.colorbar(label = u'sharpe ratio')
plt.show() 
```
