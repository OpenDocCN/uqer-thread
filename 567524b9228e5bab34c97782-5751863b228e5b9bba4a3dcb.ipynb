# 博客_生存者偏差对策略评价的影响

目前通联平台使用
universe =set_universe('HS300')  
取得的实际是当前的股票池， 
在回测时使用的也是当前的股票池，但是策略结果评价时，以沪深300指数作为参考标的。 
这会引入一定误差，但是到底会带来多少误差？（因为参与回测的股票都是“2015年12月时在沪深300标的中”，等价于使用未来信息）  
不妨测试下  
下面策略： 
capital_base = 10000000 #足够大，保证买入300个股票  
refresh_rate = 80#80工作日调整一次  
Commission(buycost=0.000, sellcost=0.000, unit='perValue')，只是为了检查和沪深300偏差，不设置手续费  
等权买入（会导致真实指数有误差，但不应该太大）  
另一个引入误差的就是买入时的整手，忽略了零头  
还有停牌无法买入，卖出引入误差
其他应该不会有什么很大的误差因素  


也就是说如果2011年就知道2015年12月时上证50的标的股票   
然后80天一调仓，等权买入sh50，可以取得12%的alpha收益，达到年化16%的绝对收益   
这12%alpha收益是没有经过任何策略优化，  
简单来说：如果一个基于sh50的策略，在区间2011-2015，没有达到16%的的年化，可以认为是失败的策略  
  
  
再试试hs300指数 
start = '2013-01-01'  ，修改为13年，这样结果会比较震撼

高达38%的年化，24%的alpha收益，，

有点超出想象，一个策略在2013-2015区间回测，如果结果达不到38%时可以认为策略失败，（如果以指数作为参考标的，毕竟指数才14%）
是否有什么没有考虑到
目前怀疑：等权买入，这个因素导致的，权重股在牛市时表现一般不如小盘股，指数编制偏重权重股占比也较大
