# 社区问答【1】：如何计算自定义技术指标的移动10日均值？

> 用户问题：

> *比如，我有下面一个语句 all_close_prices = account.get_attribute_history('closePrice',21)  然后算出一个 技术指标。然后我想再取一定时期的该指标做均线，该怎么操作？*

1. 问题分析
-----------------------
***

这是一个典型的可以使用策略全局变量可以解决的技术问题。这个问题，可以分成几个步骤：

* *如何存储存储用户自定义的技术指标？*
* *按照固定长度计算均线；*

1.1 保存至``account``
----------------

第一个小问题可以通过策略的全局变量解决，即``account``用户可以任意定义一个变量，例如``account.my_test``。那么这个``my_test``可以在``handle_data``中被持续更新。

    def initialize(account):
        account.my_test = 1
        
    def handle_data(account):
        account.my_test += 1 # 每次handle_data中，my_test会递增1
        print account._mytest

打印的结果，会是：


    1
    2
    3
    ....

1.2 计算均线
-----------------

这里用户需要需要首先计算计算自定义的"技术指标"，再计算均线。这里我们假设用户定义的"技术指标"是过去21天的**最大值**，然后计算这个最大值指标的10日均线：

首先我们在account中定义一个``list``用来保存我们计算的自定义技术指标:

    def initialize(account):
        account.price_holder = []

然后，我们在``handle_data``中计算：

    def handle_data(account)：
        all_close_prices = account.get_attribute_history('closePrice',21)
        pfyh_prices = all_close_prices['600000.XSHG'] # 获取浦发银行的过去21日价格序列
        pfyh_max = np.max(pfyh_prices)           # 计算用户自定义技术指标 - 最大值
        account.price_holder.append(pfyh_max)      # 保存至全局变量 price_holder
        pfyh_max_mean = np.mean(account.price_holder[-10:])  # 计算最近10日的最大价格均值

2. 策略的示例
--------------------
***

下面的策略例子，将上面的讲解整合起来。这个策略不做任何实际的买卖操作，只打印出相关的数值:

* 当前日期以及标的股票
* 过去21日的最大价格
* 过去10个交易日的21日最大价格序列
* 最大价格序列10日均值
