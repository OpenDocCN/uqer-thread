```python
# -------------------------------------------------
# =================================================
# 自定义函数，用于数据检测
# =================================================
import pandas as pd
import numpy as np
verify_num=1#数据检测次数初始化为1，每检测一次，次数+1，如果次数大于10，则终止检测
def verify_data(data):
        # 分割线
        print('-'*30)
        print '数据检测：'
        # 输出数据内容
        print(data)
        # 输出数据形状
        if isinstance(data,pd.DataFrame) or isinstance(data,pd.Series) or isinstance(data,np.ndarray):
            print '形状为:',data.shape
        if isinstance(data,list) or isinstance(data,dict):
            print 'len值为:',len(data)
        # 输出数据类型
        print '类型为:',type(data)
        
        # 如果次数大于1，终止检测
        global verify_num
        verify_num=verify_num+1
        if verify_num>10:
            raise ValueError('打印超过10次')
# -------------------------------------------------
```
