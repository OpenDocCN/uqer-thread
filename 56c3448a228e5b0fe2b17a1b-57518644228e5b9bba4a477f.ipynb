# 数据分析——走群众路线

```python
import matplotlib.pylab as pl

# 获取投资者情况，以周为单位，最早的数据20150504，更早的获取不到
data=DataAPI.InvestorsStatsGet(beginDate=u"20150501",endDate=u"20160201",statCycle=u"W",field=u"",pandas="1")

# 自然人与机构数量对比
data.loc[:,['endingNaturalA','endingNaturalB']].plot();
```

```python
# 自然人与机构数量相差悬殊，使用双y轴重新观察
fig=pl.figure()
ax1=fig.add_subplot(111)
ax2=ax1.twinx()
ax1.plot(data.endingNaturalA,'b',label="Natural")
ax2.plot(data.endingNaturalB,'r',label="NonNatural")
ax1.set_ylabel('Natural')
ax2.set_ylabel('Nonnatural')
pl.legend(loc='lower right')
pl.show();
```

```python
# A股交易/持股人数与开户人数比值
rateTradeA=data.endingTradingA/data.endingNaturalA
rateHoldA=data.endingHoldingA/data.endingNaturalA
fig=pl.figure()
ax1=fig.add_subplot(111)
ax2=ax1.twinx()
ax1.plot(rateTradeA,'b',label="Trade")
ax2.plot(rateHoldA,'r',label="Hold")
ax1.set_ylabel('Trade rate')
ax2.set_ylabel('Hold rate')
pl.legend(loc='upper right')
pl.show()

# 观察结果：开户人数一路增加之时，交易和持股比例持续走低，好大的坑
```

```python
# 对比持股人数比例和A股走势，看看群众路线到底对不对

# 获取上证A股指数的日行情数据
price=DataAPI.MktIdxdGet(indexID=u"",ticker=u"000002",
                         beginDate=u"20150508",endDate=u"20160201",
                         field=u"tradeDate,closeIndex",pandas="1")
# 转换为周数据，匹配投资人数据期末日期
priceW=price[price.tradeDate.isin(data.statsEnddate)]

# 作图
fig=pl.figure()
ax1=fig.add_subplot(111)
ax2=ax1.twinx()
ax1.plot(priceW.closeIndex,'b',label="Price")
ax2.plot(rateHoldA,'r',label="Hold")
ax1.set_ylabel('Index price')
ax2.set_ylabel('Hold rate')
pl.legend(loc='upper right')
pl.show()

# 观察结果：追涨和逃顶时有延迟，空仓不及时，但后期反弹时多数人视为解套机会，判断正确
# 随大流躲得了反弹后的下跌，躲不过前期的狂泻
```

```python
# 再对比下交易人数比例和A股走势
fig=pl.figure()
ax1=fig.add_subplot(111)
ax2=ax1.twinx()
ax1.plot(priceW.closeIndex,'b',label="Price")
ax2.plot(rateTradeA,'r',label="Trade")
ax1.set_ylabel('Index price')
ax2.set_ylabel('Trade rate')
pl.legend(loc='upper right')
pl.show();

# 发现交易比例更贴近走势，可惜看不到交易方向，不知道是进场买还是卖，只能说明交易活跃度紧随行情
# 换资金余额来对比，应该有用些
```

```python
# 资金余额与走势对比，观察同步程度

# 获取资金余额数据
capital=DataAPI.InvestorCapitalChgGet(beginDate=u"20150501",endDate=u"20160201",field=u"statsEnddate,capitalType,dailyAverage",pandas="1")
# 资金类型不知何解，但只有取1才合理，帮助里没解释，望高人指点
capital=capital[capital.capitalType==1]
# 匹配日期
capital=capital[capital.statsEnddate.isin(data.statsEnddate)]

# 作图
fig=pl.figure()
ax1=fig.add_subplot(111)
ax2=ax1.twinx()
ax1.plot(priceW.closeIndex,'b',label="Price")
ax2.plot(capital.dailyAverage,'r',label="Capital")
ax1.set_ylabel('Index price')
ax2.set_ylabel('Capital')
pl.legend(loc='upper right')
pl.show()

# 观察结果：除大顶没逃过，资金其他时间领先走势，在周的尺度下似乎可以用来预测走势，但遗憾没有每日数据作更细致的观察
# 可惜没有区分自然人与机构基金，否则还能看看谁嗅觉更灵敏
```
