```python
#来自量化学堂 第1课 Python入门及数据处理

#3.对上述DataFrame进行缺失值处理，缺失值按照该股票申万一级行业的均值填充。
#思路：
#1、取出某一天所有股票感兴趣因子的数据
#2、找出股票所处的申万一级行业
#3、把因子数据与行业信息merge
#4、计算每个行业的因子均值
#5、用均值对缺失值进行填充
import numpy as np
import pandas as pd
#1、取出某一天所有股票感兴趣因子的数据
trade_Date=u"20180510"
Factors_colums = ['tradeDate','secID','ticker','CTOP','DDNBT','QuickRatio']
Factors_df = DataAPI.MktStockFactorsOneDayGet(tradeDate=trade_Date,secID=u"",ticker=u"",field=Factors_colums,pandas="1")
Factors_df = Factors_df.set_index('secID',drop=True)

#2、找出股票所处的申万一级行业
#intoDate选择所取数据当天的行业分类信息
indu_columns = ['tradeDate','secID', 'ticker', 'industryName1', 'industryName2', 'industryName3'] 
indu_df = DataAPI.EquIndustryGet(industryVersionCD=u"010303", intoDate=trade_Date, field=indu_columns, pandas="1")
indu_df = indu_df.set_index('secID', drop=True)

#3、把因子数据与行业信息merge
df_merge = pd.merge(Factors_df,indu_df,on='ticker').set_index('ticker',drop=True)
tmp_df = df_merge.copy(deep=True)

#4、计算每个行业的因子均值
industry_mean = df_merge.groupby('industryName1').agg({'CTOP':'mean','DDNBT':'mean','QuickRatio':'mean'})
#industry_mean1 = df_merge.groupby('industryName1')['CTOP','DDNBT','QuickRatio'].agg(np.mean)

#5、用均值对缺失值进行填充
Factors = ['CTOP','DDNBT','QuickRatio']
for industry in industry_mean.index:
    df_merge.loc[df_merge['industryName1'] == industry,Factors] = df_merge.loc[df_merge['industryName1'] == industry,Factors].fillna(industry_mean.loc[industry][Factors])
print(len(df_merge))

#向量运算更快速
tmp_df.loc[:,Factors] = tmp_df.loc[:,Factors].fillna(tmp_df.groupby('industryName1')[Factors].transform('mean'))

df_merge.equals(tmp_df)
```
