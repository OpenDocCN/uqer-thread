```python
from functools import wraps
def wrapper_func_cache(func):
    @wraps(func)
    def wrapped(*args, **kwargs):
        if not hasattr(func, '__cache_result'):
            func.__cache_result = {}
        
        hash_key = [str(i) for i in args] + [str(j) for j in kwargs.iteritems()]
        cache_key = hash(tuple(hash_key))
        
        if cache_key not in func.__cache_result:
            func.__cache_result[cache_key] = func(*args, **kwargs)
            
        return func.__cache_result[cache_key]
    return wrapped

@wrapper_func_cache
def Monthly_const_trade_calendar(beginDate=u"20050101",endDate=u""):
    """交易日历，可以选择cache，只调用 api 一次
    """
    data = DataAPI.TradeCalGet(exchangeCD=u"XSHG,XSHE", beginDate=beginDate, endDate=endDate, field=['calendarDate', 'isMonthEnd'])
    data = data[data.isMonthEnd == 1].calendarDate.tolist()
    data = sorted(list(set(data))) 
    return data

# ST过滤
def st_remove(source_universe, st_date=None):
    """
    给定股票列表,去除其中在某日被标为ST的股票
    Args:
        source_universe (list of str): 需要进行筛选的股票列表
        st_date (datetime): 进行筛选的日期,默认为调用当天
    Returns:
        list: 去掉ST股票之后的股票列表

    Examples:
        >> universe = set_universe('A')
        >> universe_without_st = st_remove(universe)
    """
    st_date = st_date if st_date is not None else datetime.datetime.now().strftime('%Y%m%d')
    df_ST = DataAPI.SecSTGet(secID=source_universe, beginDate=st_date, endDate=st_date, field=['secID'])
    return [s for s in source_universe if s not in list(df_ST['secID'])]

# bug 函数封装问题
def winsorize_median(tmp):
    if isinstance(tmp, pd.Series):
        mc = sm.stats.stattools.medcouple(tmp)
        data = sorted(tmp)
        q1 = np.percentile(data, 25)
        q3 = np.percentile(data, 75)
        iqr = q3-q1
        if mc >= 0:
            l = q1-1.5*np.exp(-3.5*mc)*iqr
            u = q3+1.5*np.exp(4*mc)*iqr
        else:
            l = q1-1.5*np.exp(-4*mc)*iqr
            u = q3+1.5*np.exp(3.5*mc)*iqr
        tmp[tmp<l]=l
        tmp[tmp>u]=u
    return tmp   

# bug 函数封装问题
def standardize_zscore(tmp):
    if isinstance(tmp, pd.Series):
        mu = tmp.mean()
        sigma = tmp.std()
        tmp = (tmp - mu)/sigma
    return tmp   

# 获取申万一级行业分类成份股编码 
def Stock_Industry(point):
    # 申万一级指数编码
    industrys=DataAPI.IndustryGet(industryVersion=u"SW",industryVersionCD=u"",industryLevel=u"1",isNew=u"1",field=u"",pandas="1")
    indexes=list(industrys['indexSymbol'].astype(str))#行业对应的指数编码
    index_param=DataAPI.IdxGet(secID=u"",ticker=indexes,field=u"ticker,secShortName,publishDate",pandas="1")#获取指数上市时间即为对应行业开始时间
    #2014年前已存在的行业
    old_industry=index_param[index_param['publishDate']<'2014-01-01']
    old_indexes=list(old_industry['ticker'])
    old_industry_name=list(old_industry['secShortName'])
    #2014年后的所有行业
    new_industry=index_param
    new_indexes=list(new_industry['ticker'])
    new_industry_name=list(new_industry['secShortName'])
    # 申万一级指数分类函数
    SW_Old = old_indexes
    SW_New = new_indexes    
    if isinstance(point,str):
        point = pd.to_datetime(point)
    timepoint = pd.to_datetime('2014-01-01')
    if point <= timepoint:
        return SW_Old
    elif point > timepoint:
        return SW_New
    else:
        print 'wrong'

# 获取申万一级行业分类成份股       
def Industry_Constituent_Stocks(industry, timepoint):
    tmp = DataAPI.IdxConsGet(ticker=industry, intoDate=timepoint, isNew=u"",field=['consID'],pandas="1")
    return sorted(list(tmp.consID.values))
```

```python
Ddata=pd.read_csv('weight/WEIGHT.csv',index_col='date')
Ddata.stockcode=normalize_code(list(Ddata.stockcode),target_type = 'secID')
Ddata.rename(columns={'stockcode':'secID'},inplace=True)

# 获取交易日历
trade_calendar = Monthly_const_trade_calendar(beginDate=u"20110101",endDate=u"20170510")
```

```python
# t检验
start1 = time.time()
OLS_params = {}
WLS_params = {}
WLS_t_test = {}
shouyi = {}
defen = {}
w = {}
id = {}
for date in trade_calendar:
    
    # 获取当日股票池
    univ = list(Ddata.loc[date].secID)
    univ = st_remove(univ, date)

    # 获取横截面股票收益率
    stockdata   = DataAPI.MktEqudGet(tradeDate=date, secID=univ, beginDate=u"",endDate=u"",isOpen="1",field=['secID','closePrice','preClosePrice','negMarketValue'],pandas="1")
    stockdata['return'] = (stockdata.closePrice-stockdata.preClosePrice) / stockdata.preClosePrice
    stockdata['W'] = np.sqrt(stockdata.negMarketValue)
    stockdata = pd.merge(stockdata.loc[:,['secID','W','return']], Ddata.loc[date,['secID','weight']], on='secID')
  
    shouyi[date] = list(stockdata['return'].values)    
    defen[date] = list(stockdata['weight'].values)
#    defen[date] = list(standardize_zscore(winsorize_median(stockdata['weight'])))
    w[date] = list(stockdata['W'].values)
    for i in range(len(defen[date])):
        defen[date][i] = float(defen[date][i])
    # print type(defen[date])
    for i in range(len(shouyi[date])):
        shouyi[date][i] = float(shouyi[date][i])   
    # print type(shouyi[date])
    # print '股票得分字典 ' + str(len(defen[date])) + '\t股票收益字典 ' + str(len(shouyi[date])) 
    # # 股票得分字典
    # print '股票得分字典' + str(defen[date])
    # # 股票收益字典
    # print '股票收益字典' + str(shouyi[date])
    # 对股票得分和股票收益进行T检验
    df = pd.DataFrame(columns=['defen', 'shouyi'])
    df['defen'] = defen[date]
    df['shouyi'] = shouyi[date]
    X = defen[date]
    y = shouyi[date]
    W = w[date]
    # WLS回归
    wls = sm.WLS(y, X, weights=W)
    result = wls.fit()                     #获取拟合结果
    WLS_params[date] = result.params[-1]   #按date循环取回归系数(因子收益率)，最后一位存的是因子PB，前面存的是行业因子
    WLS_t_test[date] = result.tvalues[-1]  #t值
WLS = pd.Series(WLS_params)  
#OLS = pd.Series(OLS_params)

t_test = pd.Series(WLS_t_test)
print  WLS_params

print 't值序列绝对值平均值',np.sum(np.abs(t_test.values))/len(t_test)
n = [x for x in t_test.values if np.abs(x)>2]  #筛选>2的x值
print 't值序列绝对值大于2的占比——判断因子的显著性是否稳定',len(n)/len(t_test)
print 't值序列均值的绝对值除以t值序列的标准差',np.abs(t_test.mean())/t_test.std()
print '因子收益率序列平均值',WLS.mean()
print '该均值零假设检验的t值',WLS.mean()/ (WLS.std()/np.sqrt(len(WLS)-1))   #t=（x均值-u）/（方差/根号（n-1））
#输出按月的PB因子t检验结果
t_test.plot('bar',figsize=(16,5))
end1 = time.time()
print "t&收益率运行时间:%f min" %((end1 - start1)/60.0)
```

```python
# IC检验
start1 = time.time()
IC = {}
shouyi = {}
defen = {}
w = {}
id = {}
NEG = {}
#标准化
def standardize_zscore(tmp): 
    if isinstance(tmp, pd.Series):
        mu = tmp.mean()
        sigma = tmp.std()
        tmp = (tmp - mu)/sigma
    return tmp   
#去极值
def winsorize_median(tmp):
    if isinstance(tmp, pd.Series):               #isinstance（object,classinfo）判断一个对象是否是一个已知的类型
        mc = sm.stats.stattools.medcouple(tmp)   #medcouple（）用于识别非对称分布的异常值,向右倾斜的分布是正的，对于向左倾斜的分布是负的，对称分布是零
        data = sorted(tmp)                       #sorted() 函数对所有可迭代的对象进行 排序操作，返回一个新的 list
        q1 = np.percentile(data, 25)             #计算一个多维数组的任意百分比分位数的数值，25%分位数
        q3 = np.percentile(data, 75)
        iqr = q3-q1
        if mc >= 0:                              #mc表示样本偏度，当样本数据分布右偏时，提升正常数据区间上限的数值；样本数据左偏时，则会降低正常数据区间下限的数值。
            l = q1-1.5*np.exp(-3.5*mc)*iqr
            u = q3+1.5*np.exp(4*mc)*iqr
        else:
            l = q1-1.5*np.exp(-4*mc)*iqr
            u = q3+1.5*np.exp(3.5*mc)*iqr
        tmp[tmp<l]=l
        tmp[tmp>u]=u
    return tmp 

for date in trade_calendar:
    
    # 获取当日股票池
    univ = list(Ddata.loc[date].secID)
    univ = st_remove(univ, date)
    # 获取横截面股票收益率
    stockdata   = DataAPI.MktEqudGet(tradeDate=date, secID=univ, beginDate=u"",endDate=u"",isOpen="1",field=['secID','closePrice','preClosePrice','negMarketValue'],pandas="1")
    stockdata['return'] = (stockdata.closePrice-stockdata.preClosePrice) / stockdata.preClosePrice
    stockdata['fac_MV'] = standardize_zscore(winsorize_median(stockdata.negMarketValue))
    stockdata = pd.merge(stockdata.loc[:,['secID','fac_MV','return']], Ddata.loc[date,['secID','weight']], on='secID')
        
    shouyi[date] = list(stockdata['return'].values)
    defen[date] = list(stockdata['weight'].values)
    
    # 对股票得分和股票收益进行IC检验
    # NEG[date] = list(stockdata['negMarketValue'].values)
    NEG[date] = list(stockdata['fac_MV'].values)
    # X = standardize_zscore(winsorize_median(NEG[date]))
    X = NEG[date]
    # y = shouyi[date]
    y = standardize_zscore(winsorize_median(defen[date]))
    
    # OLS回归
    est=sm.OLS(y,X)
    result = est.fit()
    #print result.resid
    # print stockdata.tail()
    
    # 因子 IC 值的计算 
    Inf_coeff = stockdata.set_index('secID')
    Inf_coeff.to_excel(str(date) + 'IC_test.xlsx')
    Inf_coeff['Residual'] = result.resid
    m = Inf_coeff.loc[:,['Residual','return']].corr().iloc[0,1]
    # print m
    IC[date] = m

IC = pd.Series(IC)
print 'IC值: ', IC
print 'IC 值序列的均值大小',IC.mean(),'IC 值序列绝对值的均值大小',np.mean(np.abs(IC))
print 'IC 值序列的标准差',IC.std()
print 'IR 比率（IC 值序列均值与标准差的比值）',IC.mean()/IC.std()
n = [x for x in IC.values if x>0]
print 'IC 值序列大于零的占比',len(n)/len(IC)
m = [x for x in IC.values if np.abs(x)>0.02]  #筛选>2的x值
print 'IC值序列绝对值大于0.02的占比',len(m)/len(IC)

tmp = (IC+1.0).cumprod()-1.0  #cumprod()累积连乘
#print tmp
tmp.plot(figsize=(16,5))
end1 = time.time()
print "IC&IR运行时间:%f min" %((end1 - start1)/60.0)
```

```python
import time
from CAL.PyCAL import * 

# ------------------------------------------------- 设定回测参数 ------------------------------------------------------

start = '2016-01-01'                       # 回测起始时间
end = '2018-01-31'                         # 回测结束时间
benchmark = 'HS300'                        # 策略参考标准
universe = DynamicUniverse('HS300')           # 证券池，支持股票和基金
capital_base = 10000000                     # 起始资金
freq = 'd'                              
refresh_rate = Monthly(1)  
accounts = {
    'fantasy_account': AccountConfig(account_type='security', capital_base=10000000)
}

# ------------------------------------------------ 将因子分组并回测 ---------------------------------------------------------

# 调用quick_backtest方法
# https://uqer.datayes.com/help/api/#quick_backtest参数优化
sim_params = quartz.SimulationParameters(start, end, benchmark, universe, capital_base, refresh_rate=refresh_rate, accounts=accounts)
# 获取回测行情数据
data = quartz.get_backtest_data(sim_params)
# 运行结果
results = {}

# 遍历每一组，进行回测
for quantile_five in range(1, 6):  
    def initialize(context):                   # 初始化虚拟账户状态
        pass

    def handle_data(context): 
        account = context.get_account('fantasy_account')
        current_universe = context.get_universe('stock', exclude_halt=True)
        pre_date = context.previous_date.strftime("%Y-%m-%d")
       
        # 拿取调仓日前一个交易日的因子，并按照相应分位选择股票
        
        # 获取当日股票池
        univ = list(Ddata.loc[pre_date].secID)
        univ = st_remove(univ, pre_date)
        
        # 获取横截面股票收益率
        stockdata = DataAPI.MktEqudGet(tradeDate=pre_date, secID=univ, beginDate=u"",endDate=u"",isOpen="1",field=['secID','closePrice','preClosePrice','negMarketValue'],pandas="1")
        stockdata['return'] = (stockdata.closePrice-stockdata.preClosePrice) / stockdata.preClosePrice
        stockdata = pd.merge(stockdata.loc[:,['secID','return']], Ddata.loc[pre_date,['secID','weight']], on='secID')
        q =stockdata.sort_values('weight',axis=0, ascending=True)    #按fac因子值进行排序
        q.set_index(['secID'],inplace=True) #将股票代码设为索引

        q_min = q['weight'].quantile((quantile_five-1)*0.2)
        q_max = q['weight'].quantile(quantile_five*0.2)
        my_univ = q[(q['weight']>=q_min) & (q['weight']<q_max)].index.values
        
       # 交易部分
        positions = account.get_positions()
        sell_list = [stk for stk in positions if stk not in my_univ]
        for stk in sell_list:
            order_to(stk,0)
        
        # 在目标股票池中的，等权买入
        for stk in my_univ:
            #order_pct(stk, 1.0/len(my_univ))
            wei=float(q.loc[stk][['weight']].values[0])
            order_pct(stk, wei)
            
# ---------------策略逻辑部分结束----------------
    
    # 把回测逻辑封装到 TradingStrategy 中，供 quick_backtest 使用
    strategy = quartz.TradingStrategy(initialize, handle_data)
    # 回测部分
    bt, perf = quartz.quick_backtest(sim_params, strategy, data=data)

    # 对于回测的结果，可以通过 perf_parse 函数计算风险指标
    perf = quartz.perf_parse(bt, perf)
    
    # 保存运行结果
    t = {}
    t['bt'] = bt
    # tmp['annualized_return'] = perf['annualized_return']
    # tmp['volatility'] = perf['volatility']
    # tmp['max_drawdown'] = perf['max_drawdown']
    # tmp['alpha'] = perf['alpha']
    # tmp['beta'] = perf['beta']
    # tmp['sharpe'] = perf['sharpe']
    # tmp['information_ratio'] = perf['information_ratio']
    
    results[quantile_five] = t
    # results[quantile_five]['bt'] = bt
    # results[quantile_five]['bt'] = bt
    # results_p = pickle.dumps(results)
    # write('result_1_p.txt', results[quantile_five])
    # print 'results[' + str(quantile_five) + '] = \t' + str(results[quantile_five])
    print str(quantile_five)    #第quantile_five个分位
    
print 'done'   
```

```python
import time
from CAL.PyCAL import * 

start_time = time.time()
# ------------------------------------------------- 设定回测参数 ------------------------------------------------------

start = '2016-01-01'                       # 回测起始时间
end = '2018-01-31'                         # 回测结束时间
benchmark = 'HS300'                        # 策略参考标准
universe = DynamicUniverse('HS300')           # 证券池，支持股票和基金
capital_base = 10000000                     # 起始资金
freq = 'd'                              
refresh_rate = Monthly(1)  
accounts = {
    'fantasy_account': AccountConfig(account_type='security', capital_base=10000000)
}

# ------------------------------------------------ 将因子分组并回测 ---------------------------------------------------------

# 调用quick_backtest方法
# https://uqer.datayes.com/help/api/#quick_backtest参数优化
sim_params = quartz.SimulationParameters(start, end, benchmark, universe, capital_base, refresh_rate=refresh_rate, accounts=accounts)
# 获取回测行情数据
data = quartz.get_backtest_data(sim_params)
# 运行结果
results = {}

# 遍历每一组，进行回测
for quantile_five in range(1, 6):  
    def initialize(context):                   # 初始化虚拟账户状态
        pass

    def handle_data(context): 
        account = context.get_account('fantasy_account')
        current_universe = context.get_universe('stock', exclude_halt=True)
        pre_date = context.previous_date.strftime("%Y-%m-%d")
       
        # 拿取调仓日前一个交易日的因子，并按照相应分位选择股票
        
        # 获取当日股票池
        univ = list(Ddata.loc[pre_date].secID)
        univ = st_remove(univ, pre_date)
        
        # 获取横截面股票收益率
        stockdata = DataAPI.MktEqudGet(tradeDate=pre_date, secID=univ, beginDate=u"",endDate=u"",isOpen="1",field=['secID','closePrice','preClosePrice','negMarketValue'],pandas="1")
        stockdata['return'] = (stockdata.closePrice-stockdata.preClosePrice) / stockdata.preClosePrice
        stockdata = pd.merge(stockdata.loc[:,['secID','return']], Ddata.loc[pre_date,['secID','weight']], on='secID')
        q =stockdata.sort_values('weight',axis=0, ascending=True)    #按fac因子值进行排序
        q.set_index(['secID'],inplace=True) #将股票代码设为索引

        q_min = q['weight'].quantile((quantile_five-1)*0.2)
        q_max = q['weight'].quantile(quantile_five*0.2)
        my_univ = q[(q['weight']>=q_min) & (q['weight']<q_max)].index.values
        
       # 交易部分
        positions = account.get_positions()
        sell_list = [stk for stk in positions if stk not in my_univ]
        for stk in sell_list:
            order_to(stk,0)
        
        # 在目标股票池中的，等权买入
        for stk in my_univ:
            #order_pct(stk, 1.0/len(my_univ))
            wei=float(q.loc[stk][['weight']].values[0])
            order_pct(stk, wei)


    # 生成策略对象
    strategy = quartz.TradingStrategy(initialize, handle_data)
    
    # 开始回测, 返回说明见：https://uqer.datayes.com/help/api/#策略结果展示
    bt, perf = quartz.quick_backtest(sim_params, strategy, data=data)

    # 保存运行结果，1为因子最强组，5为因子最弱组
    results[quantile_five] = {'max_drawdown': perf['max_drawdown'], 'sharpe': perf['sharpe'], 'alpha': perf['alpha'], 'beta': perf['beta'], 'information_ratio': perf['information_ratio'], 'annualized_return': perf['annualized_return'], 'bt': bt}    

    print ('backtesting for group %s..................................' % str(quantile_five))
```

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
sns.set_style('white')

fig = plt.figure(figsize=(10,8))
fig.set_tight_layout(True)
ax1 = fig.add_subplot(211)
ax2 = fig.add_subplot(212)
ax1.grid()
ax2.grid()

for qt in results:
    bt = results[qt]['bt']

    data = bt[[u'tradeDate',u'portfolio_value',u'benchmark_return']]
    data['portfolio_return'] = data.portfolio_value/data.portfolio_value.shift(1) - 1.0   # 总头寸每日回报率
    data['portfolio_return'].ix[0] = data['portfolio_value'].ix[0]/	10000000.0 - 1.0
    data['excess_return'] = data.portfolio_return - data.benchmark_return                 # 总头寸每日超额回报率
    data['excess'] = data.excess_return + 1.0
    data['excess'] = data.excess.cumprod()                # 总头寸对冲指数后的净值序列
    data['portfolio'] = data.portfolio_return + 1.0     
    data['portfolio'] = data.portfolio.cumprod()          # 总头寸不对冲时的净值序列
    data['benchmark'] = data.benchmark_return + 1.0
    data['benchmark'] = data.benchmark.cumprod()          # benchmark的净值序列
    results[qt]['hedged_max_drawdown'] = max([1 - v/max(1, max(data['excess'][:i+1])) for i,v in enumerate(data['excess'])])  # 对冲后净值最大回撤
    results[qt]['hedged_volatility'] = np.std(data['excess_return'])*np.sqrt(252)
    results[qt]['hedged_annualized_return'] = (data['excess'].values[-1])**(252.0/len(data['excess'])) - 1.0
    ax1.plot(data['tradeDate'], data[['portfolio']], label=str(qt))
    ax2.plot(data['tradeDate'], data[['excess']], label=str(qt))
    

ax1.legend(loc=0)
ax2.legend(loc=0)
ax1.set_ylabel(u"净值", fontproperties=font, fontsize=16)
ax2.set_ylabel(u"对冲净值", fontproperties=font, fontsize=16)
ax1.set_title(u"因子不同五分位数分组选股净值走势", fontproperties=font, fontsize=16)
ax2.set_title(u"因子不同五分位数分组选股对冲后净值走势", fontproperties=font, fontsize=16)

# results 转换为 DataFrame
results_pd = pd.DataFrame(results).T.sort_index()

results_pd = results_pd[[u'alpha', u'beta', u'information_ratio', u'sharpe', u'annualized_return', u'max_drawdown',  
                         u'hedged_annualized_return', u'hedged_max_drawdown', u'hedged_volatility']]

cols = [(u'风险指标', u'Alpha'), (u'风险指标', u'Beta'), (u'风险指标', u'信息比率'), (u'风险指标', u'夏普比率'), (u'纯股票多头时', u'年化收益'),
        (u'纯股票多头时', u'最大回撤'), (u'对冲后', u'年化收益'), (u'对冲后', u'最大回撤'), (u'对冲后', u'收益波动率')]
results_pd.columns = pd.MultiIndex.from_tuples(cols)
results_pd.index.name = u'五分位组别'
results_pd
```
