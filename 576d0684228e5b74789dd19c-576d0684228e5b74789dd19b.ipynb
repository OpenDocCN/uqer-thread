# 基金的收益和基金基金经理
我持两者不能随便分开的观点。这边我会简单介绍下如何去保证基金的经理没换过
# 基金的短期收益不能代表基金经理的能力很有可能是运气
所以短于三年的基金，我不看。三年只是一个臆断的时期。喜欢统计的可以自己去具体算算自己想要的指标。

```python
import pandas as pd
import datetime

# 整体打印的话，不难看到下面的有错误数据，于是我们又多了一步preprocessing的步骤
fund_E = DataAPI.FundGet(category='E',listStatusCd='L',operationMode='O')

# fund_E[fund_E['secShortName'].str.contains('量化')]，早期太多的跟踪指数的基金了。这一句只是事后为了挑选个不是跟踪指数的。

fund_E= fund_E[fund_E['secID'].str.match('[0-9]')]
fund_E.head(10)
```

```python
'''以下为有用的指标，我们要根据相应的情况过滤过。第一只需要分析上市基金就行了；第二上市日期要早于目前三年
managerName	str	基金经理
listDate	str	上市日期
'''
fund_E['listDate'] = fund_E['listDate'].apply(pd.to_datetime)
```

```python
lag = pd.DateOffset(months=36)
current = datetime.datetime.now()
fund_E = fund_E[fund_E['listDate'] < current - lag]
```

DataAPI.FundManagerGet
获取某只基金从成立以来，基金经理与基金经理助理的任职变动情况，包含基金经理的姓名，学历、性别、出生年、证书、入职时间和离职时间等。数据更新频率为不定期。

```python
tickerlist = ','.join(fund_E['ticker'].values)
tickerlist
```

```python
DataAPI.FundManagerGet(ticker = tickerlist)
```

上面这个实际上已经可以筛选出是否发生经理人变更了。因为只要确定每只基金最早的一个经理人在就行了。有些直接是跟指数挂钩的，不要管它有没有换人。但是一旦涉及到基金经理这个问题的时候，马上与他管理的基金数等各种可能因素要结合，我就不说下去了。并且再说下去可能会失之广告的嫌疑。
