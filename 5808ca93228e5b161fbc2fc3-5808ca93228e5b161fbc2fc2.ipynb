# `策略逻辑`
***
  
- **接收信息**：每个调仓周期更新一次信息，得到新的行情、技术指标等数据；  


- **预测收益**：利用得到的信息对市场下周收益率情况进行预测，本篇简单采用10日移动平均收益率作为之后收益率的估计；


- **配置权重**：
    - 根据预测收益配置下期资产权重，在这里使用passive-aggressive的方法，即在满足一定预测组合收益的情况下最小化权重变动：
\[{min} \ \    \frac12||b_t - b_{t-1}||^2\\  
s.t. \ \ \ \  {b_t}^T*\widehat{r_{t}}\ge\epsilon\\
\ \ \ \ \ \ \ \ \ \ 1^T*{b_t} =1\\
\ \ \ \ \ \ \ \ \ \ {b_t}\ge0\]
	- 这一逻辑被称为passive-aggressive是因为策略会随着$\epsilon$变化而改变风格，$\epsilon$较大时，策略调仓较多，且持仓相对集中，显得aggressive，$\epsilon$较小时则相反。
    
    
- **风险控制**：这里采用比较简易的风险控制方法，当组合收益相较于参考价回撤超过5%时，空仓几个调整期间。
