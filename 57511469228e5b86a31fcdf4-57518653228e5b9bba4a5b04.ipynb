```python
# 资金流
df = DataAPI.MktEquFlowOrderGet(secID='601933.XSHG', beginDate='20160101', endDate='20160531', 
                                field=['ticker', 'secID', 'tradeDate','netInflowS', 'netInflowM', 'netInflowL', 'netInflowXl'])
df['netIn'] = df.netInflowS + df.netInflowM + df.netInflowL + df.netInflowXl

df2= DataAPI.MktEqudAdjGet(secID='601933.XSHG', beginDate='20160101', endDate='20160531',
                           field=['ticker', 'tradeDate', 'openPrice', 'closePrice', 'preClosePrice'])
df2['change'] = (df2.closePrice - df2.openPrice) / df2.preClosePrice

df3 = df.merge(df2, on=['ticker', 'tradeDate'])

print u'601933 资金流入流出与当日涨跌数据'
print df3.head().to_html()

df3['netIn_multi_change'] = df3.netIn * df3.change
ax = df3.plot(x=['tradeDate'], y=['netIn_multi_change'], kind='bar', figsize=(20, 10), title=u'601933 当日资金流流向与涨跌乘积')
share_0602.fig_style(ax, [u'资金流净流入 * 当日涨跌幅'], legend_loc='best')
```
