提出几个问题：
（1）:相关性大的代表走势相同（这里把量纲的影响消除了，因为归一化）但是如上图所示，我们选出了相关性最高的两只股票，但是他们的价差序列含有的趋势性较强
，或者说当这两只股票同时涨跌同样的幅度时他们的价差实际上变大了，并且会越来越大（当两只股票的走势仍按之前的走势），当两只股票走势相反时价差又向反方向走，这样对套利极其不利。
(2):第一种改进方式就是利用协方差排序，（不用相关系数），第二种就是把价差序列进行检测选出我们想要的序列（围绕均值上下波动，均值较平稳）。

第一种改进的实验

```python
# 计算协方差 （包含量纲）
covmat = np.cov(allclsprx, y=None, rowvar=1, bias=0, ddof=None)
d = len(covmat)
covlist = []
for i in range(d-1):
    for j in range(i+1, d):
        covlist.append((covmat[i, j], i, j))#每个元素是一个tuple 
print covlist
# 挑出负相关系数与正相关系数最大的各10只股票
neglist = nsmallest(10, covlist, key=lambda x:x[0])
neglist = [map(idxmap.get, p[1:])+[p[0]] for p in neglist]
poslist = nlargest(10, covlist, key=lambda x:x[0])
poslist = [map(idxmap.get, p[1:])+[p[0]] for p in poslist]
```
