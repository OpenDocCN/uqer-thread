兴业证券：基本面量化视角下的机构持仓信息研究系列之二：基金优选下的重仓股信息研究 

原文策略分为两部分，第一部分是从混合及股票基金中选择优秀基金，第二部分是从优秀基金持仓股中选择重仓股。

本文先复现第一部分：选择优秀基金。

# 样本选择：
普通股票型和偏股混合基金
成立1年3个月以上，
最近规模>5亿，
最近4季度平均股票仓位>=60%（只考虑选股能力）
最近4季度重仓股平均更换比例<=70%
平均更换比例=当前季度前十大重仓股相比上季度重仓股的数量变动比例。即当前季度与上季度重仓股不同的数量/10
近1年基金经理未更换(未实现)
# 因子合成
选股Alpha*0.3+最大回撤*0.2+基金份额*0.2+机构持有比例*0.2+基金从业人员持有比例*0.2
原文中Alpha使用各种风格指数回归后计算出的选股Alpha，这里偷懒就直接用优矿的alpha数据代替了。

```python
from datetime import *
import numpy as np
import pandas as pd
def rebalance(tradeDate):
    #==================样本筛选==========================================================================#
    #funds=DataAPI.IdxConsGet(ticker=u"930950",intoDate=tradeDate,field=u"consID,consShortName",pandas="1")
    funds1=DataAPI.FundGet(category=u"H",operationMode=u"O",endDate=tradeDate,field=u"",pandas="1")
    #去掉指数基金和指数增强基金
    funds1=funds1[~ ((funds1['indexFund']=='I') | (funds1['indexFund']=='EI'))]
    funds1=funds1[funds1['isQdii']!=1]
    funds1=funds1[(funds1['secID'].str.contains('OFCN')) | (funds1['secID'].str.contains('XSHG')) | (funds1['secID'].str.contains('XSHE'))]
    #成立时间大于1年零3个月
    days=15*30
    deltaDay=timedelta(days=days)
    start_date=(datetime.strptime(tradeDate,'%Y%m%d')-deltaDay).strftime("%Y-%m-%d")
    funds1=funds1[funds1.establishDate<start_date]
    secID=list(funds1['secID'])
    #print(secID)
    #最近规模大于5亿
    days=3*30
    deltaDay=timedelta(days=days)
    start_date=(datetime.strptime(tradeDate,'%Y%m%d')-deltaDay).strftime("%Y%m%d")
    data=DataAPI.FundAssetsGet(secID=secID,beginDate=start_date,endDate=tradeDate,
                               field=u"secID,netAsset,reportDate",pandas="1")
    secID=list(data[data.netAsset>5e+08]['secID'])
    #最近四季度股票占比大于60%
    days=12*30
    deltaDay=timedelta(days=days)
    start_date=(datetime.strptime(tradeDate,'%Y%m%d')-deltaDay).strftime("%Y%m%d")
    data=DataAPI.FundAssetsGet(secID=secID,beginDate=start_date,endDate=tradeDate,
                               field=u"secID,equityRatioInTa",pandas="1")
    data=data.groupby(['secID']).mean()
    data=data[data['equityRatioInTa']>60]
    secID=list(data.index)
    #最近4个季度重仓股变动比例<=70%
    data=DataAPI.FundHoldingsGet(secID=secID,beginDate=start_date,endDate=tradeDate,secType="E"
                                 ,field=u"secID,reportDate,holdingSecID,marketValue",pandas="1")
    data.sort_values(by='marketValue',inplace=True,ascending=False)
    data=data.groupby(['secID','reportDate']).head(10)
    
    # 对比两个季度持股差别
    change={}
    for sec in secID:
        tmpData=data[data['secID']==sec]
        reportDate=list(tmpData['reportDate'].drop_duplicates())
        result=[]
        for i in range(1,len(reportDate)):
            old_stock=list(tmpData[tmpData['reportDate']==reportDate[i-1]]['holdingSecID'])
            new_stock=list(tmpData[tmpData['reportDate']==reportDate[i]]['holdingSecID'])
            result.append(len(set(old_stock)&set(new_stock)))
        change[sec]=np.mean(result)
    change=pd.Series(change)
    secID=list(change[change>3].index)
    #==================因子合成==========================================================================#
    #alpha,最大回撤
    data1=DataAPI.FundPerfIndicGet(dataDate=tradeDate,secID=secID,window=u"5",field=u"secID,maxDD,realizedAlpha",pandas="1")
    data1.set_index('secID',inplace=True)
    data2=DataAPI.FundHolderInfoGet(secID=secID,beginDate=start_date,
                                    endDate=tradeDate,
                                    field=u"secID,reportDate,holderNumber,avgHoldShares,instHoldRatio,pracHoldRatio",
                                    pandas="1")
    #取最近一次数据
    end_date=data2['reportDate'].max()
    data2=data2[data2['reportDate']==end_date]
    data2.set_index('secID',inplace=True)
    data2.drop('reportDate',inplace=True,axis=1)
    data=pd.merge(data1,data2,left_index=True,right_index=True)
    data['holding']=data['holderNumber']*data['avgHoldShares']#计算份额
    data.drop(['holderNumber','avgHoldShares'],inplace=True,axis=1)#去掉不需要的因子
    data['maxDD']=-1*data['maxDD']#最大回撤重新排序
    #计算标准分
    data_result=(data-data.mean())/data.std()
    #合成因子
    data_result['result']=data_result['realizedAlpha']*0.3+data_result['maxDD']*0.1+data_result['holding']*0.2+data_result[
        'instHoldRatio']*0.2+data_result['pracHoldRatio']*0.2
    data_result=data_result.sort_values('result',ascending=False)
    return list(data_result.iloc[:5].index)
```

```python
portfolio={}
init_value=10000
portfolio[dates[0].replace('-','')]=init_value
is_first=True#第一个日期不计算收益
volumn=''#每只基金持仓数量
holding=''
for d in dates:
    d=d.replace('-','')
    if is_first:
        is_first=False
        holding=rebalance(d)
        print(d,holding)
        for h in holding:
            per_value=init_value/len(holding)#每只基金金额
            #当前基金净值
            price=DataAPI.FundNavGet(secID=holding,dataDate=d,field=u"secID,ADJUST_NAV",pandas="1")
            price.set_index('secID',inplace=True)
            volumn=per_value/price
            volumn.columns=['volumn']
    else:
        #当前基金净值
        price=DataAPI.FundNavGet(secID=holding,dataDate=d,field=u"secID,ADJUST_NAV",pandas="1")
        price.set_index('secID',inplace=True)
        data=pd.merge(price,volumn,left_index=True,right_index=True)
        data['value']=data['ADJUST_NAV']*data['volumn']
        total_value=data['value'].sum()
        portfolio[d]=total_value
        #每年4月最后一个交易日轮仓
        if d[5] in ['4']:
            holding=rebalance(d)
            print(d,holding)
            per_value=total_value/len(holding)
            price=DataAPI.FundNavGet(secID=holding,dataDate=d,field=u"secID,ADJUST_NAV",pandas="1")
            price.set_index('secID',inplace=True)
            volumn=per_value/price
            volumn.columns=['volumn']
        
```
