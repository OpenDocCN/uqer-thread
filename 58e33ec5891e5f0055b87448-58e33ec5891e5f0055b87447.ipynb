优化3：线性回归系数
思路：按照动量来评估未来一段时间收益，其本质基于“价格趋势继续“的假设。所以，最能体现价格趋势是回归拟合价格信息的系数。此系数代替原来的动量，进行排名筛选，为了简单，线性回归的1次方系数代替动量
