# 【从零开始学uqer】ATR的吐槽

	新的一周，新的启程，知道ATR是什么吗？

	ATR(平均真实波幅,即average true range)，优矿因子库里描述ATR6：6日均幅指标（Average TRUE Ranger），取一定时间周期内的股价波动幅度的移动平均值，是显示市场变化率的指标，主要用于研判买卖时机。属于超买超卖型因子。
    
    ATR是韦尔德开发的，它给交易者一种感觉，就是知道什么是历史波幅，使他们在真实的市场中为交易做好准备。
    
	真实波幅是下列3个等式的最大值：
	1. TR = H – L
	2. TR = H – Cl
	3. TR = Cl – L
	
	TR-真实波幅
	H-今日最高值
	L-今日最低值
    CL-前一日的收盘价
    
    公式很好理解，作为一名业余非专业股票研究爱好者，抱着严谨的态度，研究下因子数据

```python
def st_remove(source_universe, st_date=None):
    """
    给定股票列表,去除其中在某日被标为ST的股票
    Args:
        source_universe (list of str): 需要进行筛选的股票列表
        st_date (datetime): 进行筛选的日期,默认为调用当天
    Returns:
        list: 去掉ST股票之后的股票列表

    Examples:
        >> universe = set_universe('A')
        >> universe_without_st = st_remove(universe)
    """
    st_date = st_date if st_date is not None else datetime.datetime.now().strftime('%Y%m%d')
    df_ST = DataAPI.SecSTGet(secID=source_universe, beginDate=st_date, endDate=st_date, field=['secID'])
    return [s for s in source_universe if s not in list(df_ST['secID'])]
```

```python
begin_date = datetime.strptime('2009-01-01', '%Y-%m-%d')    # 起始时间
end_date = datetime.strptime('2016-1-1', '%Y-%m-%d')      # 截止时间
universe = set_universe('A') #获取股票
```

```python
# 计算各股票历史区间前瞻回报率，未来window天的收益率
def getForwardReturnsAll(universe, begin, end, window, file_name):
    print str(window) + ' days forward returns will be calculated for ' + str(len(universe)) + ' stocks:'
    count = 0
    secs_time = 0
    start_time = time.time()#时间计数，这个时间计数可以评估工作量，简直好评
    
    #数据采集
    ret_data = pd.DataFrame()
    N = len(universe)
    
    for stk in universe:
        data = DataAPI.MktEqudAdjGet(secID=stk, beginDate=begin, endDate=end, field='secID,tradeDate,closePrice')
        tmp_ret_data = data.sort('tradeDate')
        
        tmp_ret_data['forwardReturns'] = tmp_ret_data['closePrice'].shift(-window) / tmp_ret_data['closePrice']  - 1.0
        tmp_ret_data['forwardReturns'] = tmp_ret_data['forwardReturns']/window
        tmp_ret_data = tmp_ret_data[['tradeDate','forwardReturns']]
        tmp_ret_data.columns = ['tradeDate', stk]

        if ret_data.empty:
            ret_data = tmp_ret_data
        else:
            ret_data = ret_data.merge(tmp_ret_data, on='tradeDate', how='outer')
        
        # 打印进度部分
        count += 1
        if count > 0 and count % N == 0:
            finish_time = time.time()
            print count,
            print '  ' + str(np.round((finish_time-start_time) - secs_time, 0)) + ' seconds elapsed.'
            secs_time = (finish_time-start_time)#时间计数
    
    ret_data.to_csv(file_name)
    return ret_data

# 获取股票历史
def get_factor(universe, begin, end, file_name):
    print  'ATR will be calculated for ' + str(len(universe)) + ' stocks:'
    count = 0
    secs_time = 0
    start_time = time.time()#时间计数
    
    #数据采集
    N = len(universe)
    ret_data = pd.DataFrame()
    
    for stk in universe:
        data = DataAPI.MktStockFactorsDateRangeGet(secID=stk, beginDate=begin, endDate=end, field='secID,tradeDate,ATR6,ATR14')    # 拿取数据
        tmp_ret_data = data.sort('tradeDate')
        
        tmp_ret_data['ATR'] = tmp_ret_data['ATR6']
        tmp_ret_data = tmp_ret_data[['tradeDate','ATR']]
        tmp_ret_data.columns = ['tradeDate', stk]

        if ret_data.empty:
            ret_data = tmp_ret_data
        else:
            ret_data = ret_data.merge(tmp_ret_data, on='tradeDate', how='outer')

        # 打印进度部分
        count += 1
        if count > 0 and count % N == 0:
            finish_time = time.time()
            print count,
            print '  ' + str(np.round((finish_time-start_time) - secs_time, 0)) + ' seconds elapsed.'
            secs_time = (finish_time-start_time)#时间计数
    
    ret_data.to_csv(file_name)
    return ret_data
```

# 因子处理

```python
def winsorize_standardize_series(se):#去极值+标准化
    q = se.quantile([0.025, 0.975])
    if isinstance(q, pd.Series) and len(q) == 2:
        se[se < q.iloc[0]] = q.iloc[0]
        se[se > q.iloc[1]] = q.iloc[1]
    se_std = se.std()
    se_mean = se.mean()
    se_standardize = (se - se_mean)/se_std
    return se_standardize
```

```python
#ATR因子处理         
print '======================='
start_time = time.time()
factor_data_5D_deal = factor_data_5D.apply(winsorize_standardize_series,axis=1)
factor_data_5D_deal.to_csv('ATR_deal.csv')
finish_time = time.time()
print ''
print str(finish_time-start_time) + ' seconds elapsed in total.'
```

```python
# 因子历史表现
n_quantile = 10
# 统计十分位数
cols_mean = ['meanQ'+str(i+1) for i in range(n_quantile)]
cols = cols_mean
corr_means = pd.DataFrame(index=factor_data_5D_deal.index, columns=cols)

# 计算相关系数分组平均值
for dt in corr_means.index:
    qt_mean_results = []

    # 相关系数去掉nan和绝对值大于1的
    tmp_factor = factor_data_5D_deal.ix[dt].dropna()
    #tmp_factor = tmp_factor[(tmp_factor<=1.0) & (tmp_factor>=-1.0)]
    
    pct_quantiles = 1.0/n_quantile
    for i in range(n_quantile):
        down = tmp_factor.quantile(pct_quantiles*i)
        up = tmp_factor.quantile(pct_quantiles*(i+1))
        mean_tmp = tmp_factor[(tmp_factor<=up) & (tmp_factor>=down)].mean()
        qt_mean_results.append(mean_tmp)
    corr_means.ix[dt] = qt_mean_results


# ------------- 因子历史表现作图 ------------------------

fig = plt.figure(figsize=(12, 6))
ax1 = fig.add_subplot(111)

lns1 = ax1.plot(corr_means.index, corr_means.meanQ1, label='Q1')
lns2 = ax1.plot(corr_means.index, corr_means.meanQ5, label='Q5')
lns3 = ax1.plot(corr_means.index, corr_means.meanQ10, label='Q10')

lns = lns1+lns2+lns3
labs = [l.get_label() for l in lns]
ax1.legend(lns, labs, bbox_to_anchor=[0.5, 0.1], loc='', ncol=3, mode="", borderaxespad=0., fontsize=12)
ax1.set_ylabel(u'因子', fontproperties=font, fontsize=16)
ax1.set_xlabel(u'日期', fontproperties=font, fontsize=16)
ax1.set_title(u"因子历史表现", fontproperties=font, fontsize=16)
ax1.grid()
```

```python
#分组超额收益分析
print '======================='
start_time = time.time()

n_quantile = 40
# 统计十分位数
cols_mean = [i+1 for i in range(n_quantile)]
cols = cols_mean

excess_returns_means_h = pd.DataFrame(index=factor_h.index, columns=cols)
excess_returns_median_h = pd.DataFrame(index=factor_h.index, columns=cols)

# 计算ILLIQ分组的超额收益平均值
for dt in factor_h.index:
    qt_mean_results = []
    qt_median_results = []
    
    # ILLIQ去掉nan
    tmp_factor = factor_h.ix[dt].dropna()
    tmp_return = forward_10d_return_data.ix[dt].dropna()
    tmp_return_mean = tmp_return.mean()
    tmp_return_median = tmp_return.median()
    
    pct_quantiles = 1.0/n_quantile
    for i in range(n_quantile):
        down = tmp_factor.quantile(pct_quantiles*i)
        up = tmp_factor.quantile(pct_quantiles*(i+1))
        i_quantile_index = tmp_factor[(tmp_factor<=up) & (tmp_factor>=down)].index
        mean_tmp = tmp_return[i_quantile_index].mean() - tmp_return_mean
        median_tmp = tmp_return[i_quantile_index].median() - tmp_return_median
        qt_mean_results.append(mean_tmp)
        qt_median_results.append(median_tmp)
        
    excess_returns_means_h.ix[dt] = qt_mean_results
    excess_returns_median_h.ix[dt] = qt_median_results

excess_returns_median_h.dropna(inplace=True)
excess_returns_means_h.dropna(inplace=True)

#====================================================

excess_returns_means_l = pd.DataFrame(index=factor_l.index, columns=cols)
excess_returns_median_l = pd.DataFrame(index=factor_l.index, columns=cols)

# 计算ILLIQ分组的超额收益平均值
for dt in factor_l.index:
    qt_mean_results = []
    qt_median_results = []
    
    # ILLIQ去掉nan
    tmp_factor = factor_l.ix[dt].dropna()
    tmp_return = forward_10d_return_data.ix[dt].dropna()
    tmp_return_mean = tmp_return.mean()
    tmp_return_median = tmp_return.median()
    
    pct_quantiles = 1.0/n_quantile
    for i in range(n_quantile):
        down = tmp_factor.quantile(pct_quantiles*i)
        up = tmp_factor.quantile(pct_quantiles*(i+1))
        i_quantile_index = tmp_factor[(tmp_factor<=up) & (tmp_factor>=down)].index
        mean_tmp = tmp_return[i_quantile_index].mean() - tmp_return_mean
        median_tmp = tmp_return[i_quantile_index].median() - tmp_return_median
        qt_median_results.append(median_tmp)
        qt_mean_results.append(mean_tmp)
        
    excess_returns_means_l.ix[dt] = qt_mean_results
    excess_returns_median_l.ix[dt] = qt_median_results

excess_returns_means_l.dropna(inplace=True)
excess_returns_median_l.dropna(inplace=True)

finish_time = time.time()
print ''
print str(finish_time-start_time) + ' seconds elapsed in total.'
```

```python
def st_remove(source_universe, st_date=None):
    """
    给定股票列表,去除其中在某日被标为ST的股票
    Args:
        source_universe (list of str): 需要进行筛选的股票列表
        st_date (datetime): 进行筛选的日期,默认为调用当天
    Returns:
        list: 去掉ST股票之后的股票列表

    Examples:
        >> universe = set_universe('A')
        >> universe_without_st = st_remove(universe)
    """
    st_date = st_date if st_date is not None else datetime.datetime.now().strftime('%Y%m%d')
    df_ST = DataAPI.SecSTGet(secID=source_universe, beginDate=st_date, endDate=st_date, field=['secID'])
    return [s for s in source_universe if s not in list(df_ST['secID'])]
```

```python
#根据bt分析持股情况
def get_mv_and_ind(bt):
    report = bt[['tradeDate','security_position']]
    date_lcap = {}
    universe_ind = []
    for i in range(len(report)):
        df_mv = DataAPI.MktStockFactorsOneDayGet(tradeDate=report['tradeDate'].iloc[i].strftime('%Y%m%d'),
                                         secID=report['security_position'].iloc[i].keys(),ticker=u"",field=u"ticker,LCAP",pandas="1")
        df_mv_total = DataAPI.MktStockFactorsOneDayGet(tradeDate=report['tradeDate'].iloc[i].strftime('%Y%m%d'),
                                         secID=u"",ticker=u"",field=u"ticker,LCAP",pandas="1")
        df_mv_total = df_mv_total.dropna()
        df_mv_quantile_25 = df_mv_total['LCAP'].quantile(0.25)
        df_mv_quantile_75 = df_mv_total['LCAP'].quantile(0.75)
        df_ind = DataAPI.EquIndustryGet(industryVersionCD=u"010303",industry=u"",secID=report['security_position'].iloc[i].keys(),ticker=u"",
                           intoDate=report['tradeDate'].iloc[i].strftime('%Y%m%d'),field=u"industryName1",pandas="1")
        universe_ind = universe_ind + df_ind['industryName1'].tolist()
        date_lcap[report['tradeDate'].iloc[i]] =  [df_mv['LCAP'].mean(),df_mv_quantile_25,df_mv_quantile_75]
    return date_lcap,universe_ind 

def series_mv_plot(date_lcap):
    date_lcap_df = pd.DataFrame(date_lcap).T
    fig = plt.figure(figsize=(18,7))
    ax = fig.add_subplot(111)
    ax.plot(date_lcap_df.index,date_lcap_df[0].values)
    ax.plot(date_lcap_df.index,date_lcap_df[1].values,'--')
    ax.plot(date_lcap_df.index,date_lcap_df[2].values,'r--')
    ax.legend([u"策略持仓股票平均市值",u"全市场股票市值25%分位数",u"全市场股票市值75%分位数"],loc='upper left',bbox_to_anchor=(0.01,0.98),prop=font,frameon= False) 
    
def get_ind_precent_plot(universe_ind):
    universe_ind_df = pd.DataFrame(universe_ind)
    ind_percent = universe_ind_df[0].value_counts() / len(universe_ind_df)
    fig = plt.figure(figsize=(18,7))
    ax = fig.add_subplot(111)
    ind_percent.plot(kind='bar',ax=ax,color='skyblue')
    def myfunc(x, pos=0): 
        return "{0:.2f}%".format(x * 100)
#    ax.yaxis.set_major_formatter(FuncFormatter(myfunc))
    ax.set_xticklabels([item.decode('utf-8') for item in  ind_percent.index],fontproperties=font)
    ax.spines['top'].set_color('none')
    ax.xaxis.set_ticks_position('bottom')
    ax.set_title(u"持仓行业分布", fontproperties=font, fontsize=18)
    plt.show() 
```
