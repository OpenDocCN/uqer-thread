# 1.何为跨品种套利

从百度百科的定义来看，跨品种套利是指在买入（或卖出）某种期货合约的同时卖出或买入相关的另一种商品（合约），当两者的价差或价比收缩或扩大至一定程度时，平仓了结的交易方式。它的逻辑在于寻找不同品种，但是具有一定相关性的商品期货间相对稳定的价格关系，以期这种价格关系从偏离回到正常过程中价格波动所带来的利润。跨品种套利所选择的品种从经济角度来说，它们可能处于同一产业链的上下游，比如玉米与淀粉，也有可能具有可替代或者互补的关系，比如豆油与菜油；从统计角度来看，跨品种套利所选择的品种的价格具有很强的相关性并且满足协整的关系。这里只是以股指期货为例（现实中的股指期货的交易存在一些限制），给出一个简单的跨品种套利的例子。

# 2.上证50指数与沪深300指数

根据统计，上证50指数与沪深300指数高度相关，相关系数达到了99%，这表明两者价格的趋同性使得这两个品种间的套利成为了可能。下图直观的展示了两指数每日的走势，可以看到高度一致。

```python
import DataAPI
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
from CAL.PyCAL import font

# 获取数据
sz50 = DataAPI.MktIdxdGet(ticker=u"000016",beginDate=u"20150101",endDate=u"20160601",field=[u"closeIndex", u"tradeDate"],pandas="1")
hs300 = DataAPI.MktIdxdGet(ticker=u"000300",beginDate=u"20150101",endDate=u"20160601",field=[u"closeIndex", u"tradeDate"],pandas="1")

# 计算相关系数
rho = np.corrcoef(sz50['closeIndex'], hs300['closeIndex'])
print rho[0][1]

# 绘制走势图
fig = plt.figure(figsize=(10,4))
fig.set_tight_layout(True)
ax = fig.add_subplot(111)
ax.grid()
x_axis = np.array(sz50['tradeDate']).tolist()
time = []
for i in range(len(x_axis)):
    time.append(datetime.strptime(x_axis[i], "%Y-%m-%d"))
y_axis1 = np.array(sz50['closeIndex']).tolist()
y_axis2 = np.array(hs300['closeIndex']).tolist()
ax.plot(time, y_axis1, '-r', label='IHM0')
ax.plot(time, y_axis2, '-g', label='IFM0')
ax.legend(loc=2)
ax.set_title(u'价格走势图', fontproperties=font, fontsize=16)
```

# 3.协整检验

分别对上证50指数与沪深300指数进行单位根检验

```python
from statsmodels.tsa.stattools import adfuller
test = adfuller(np.array(y_axis1))
result1 = pd.Series(test[0:4], index=['Test Statistic','p-value','Lags Used','Number of Observations Used'])
for key,value in test[4].items():
    result1['Confidence level (%s)'%key] = value
print pd.DataFrame(result1, columns=['ratio'])

test = adfuller(np.array(y_axis2))
result2 = pd.Series(test[0:4], index=['Test Statistic','p-value','Lags Used','Number of Observations Used'])
for key,value in test[4].items():
    result2['Confidence level (%s)'%key] = value
print pd.DataFrame(result2, columns=['ratio'])
```

均接受原假设，表示上证50与沪深300指数序列不平稳，对其进行一阶差分后再做单位根检验：

```python
from statsmodels.tsa.stattools import adfuller
test = adfuller(np.array(y_axis1)[1:] - np.array(y_axis1)[:-1])
result1 = pd.Series(test[0:4], index=['Test Statistic','p-value','Lags Used','Number of Observations Used'])
for key,value in test[4].items():
    result1['Confidence level (%s)'%key] = value
print pd.DataFrame(result1, columns=['ratio'])

test = adfuller(np.array(y_axis2)[1:] - np.array(y_axis2)[:-1])
result2 = pd.Series(test[0:4], index=['Test Statistic','p-value','Lags Used','Number of Observations Used'])
for key,value in test[4].items():
    result2['Confidence level (%s)'%key] = value
print pd.DataFrame(result2, columns=['ratio'])
```

可以看出进行一阶差分后我们得到了两个平稳的序列

```python
# 拟合回归方程
import scipy.stats as stats
result = stats.linregress(y_axis1, y_axis2)

# 获得残差并绘制残差图
error = np.array(y_axis2)-result[0]*np.array(y_axis1)-result[1]
fig = plt.figure(figsize=(10,2))
fig.set_tight_layout(True)
ax = fig.add_subplot(111)
ax.grid()
ax.plot(error)
ax.set_title(u"残差序列图", fontproperties=font, fontsize=16)
```

```python
#对残差进行单位根检验
test = adfuller(error)
result2 = pd.Series(test[0:4], index=['Test Statistic','p-value','Lags Used','Number of Observations Used'])
for key,value in test[4].items():
    result2['Confidence level (%s)'%key] = value
print pd.DataFrame(result2, columns=['ratio'])
```

从指数的结构来看，沪深300的成分股中金融地产板块占比最大，约为[40%](http://finance.sina.com.cn/qizhi/hs300.html#gxzs)。而上证50指数挑选的是上海证券市场流动性最强，规模最大的50支具有代表性的股票组成，其中金融板块占比超过65%。所以直观来说，在金融板块表现较好时可以买入上证50指数，卖出沪深300指数套利，而在金融板块弱势的情况下卖出上证50指数，买入沪深300指数套利。
