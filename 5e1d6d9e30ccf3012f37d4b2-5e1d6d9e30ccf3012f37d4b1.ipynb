```python
import numpy as np
import pandas as pd
from CAL.PyCAL import *
from datetime import date

cal = Calendar('China.SSE')

def myTradeDate(predays="-1B",fmt="%Y%m%d"):
    """
    获取交易日期
    eg. 
    # 获取倒数第20个交易日
    # myTradeDate(20)
    """
    return cal.advanceDate(date.today(), predays,  BizDayConvention.Following).strftime(fmt)
def extKCB(secIDs):
    """
    剔除科创板股票
    """
    tmp=DataAPI.EquGet(secID=secIDs,equTypeCD=u"A",ListSectorCD=u"4",field=u"secID",pandas="1")
    return list(set(secIDs)-set(tmp.secID.values))
def universeTicker(extST=True, extBaolei=False):
    """
    剔除科创板股票
    """
    tmp=DataAPI.EquGet(equTypeCD=u"A",ListSectorCD=u"1,2,3",field=u"ticker",pandas="1")
    ticker=tmp.ticker.values
    #去除停牌的股票
    tradeDate=myTradeDate('-1B')
    if extST:
        st=DataAPI.SecSTGet(ticker=ticker,beginDate=tradeDate,endDate=tradeDate,field=u"ticker",pandas="1")['ticker'].values
        ticker=list(set(tmp.ticker.values)-set(st))
    if extBaolei:
        ticker=extGoodWillTicker(ticker,False,False)
    return ticker
def goodWillBaolei(secIDs,withPrint=True):
    """
    找出并打印2019年3季报商誉存雷的股票
    eg.
    # goodWillBaolei(set_universe("HS300"))
    """
    thisYear="2019"
    # 合并资产负债表(3季度)
    data = DataAPI.FdmtBSGet(secID=secIDs,reportType=u"Q3",beginYear=thisYear,endYear=thisYear,
                             field=u"secID,goodwill,TAssets,TLiab,minorityInt",pandas="1")
    data=data.dropna()
    data=data.drop_duplicates()
    # 净利润表（3季度）
    field=u"secID,NIncomeAttrP,secShortName"
    data1=DataAPI.FdmtISGet(secID=secIDs,reportType=u"CQ3",beginYear=thisYear,endYear=thisYear,field=field,pandas="1")

    data2=data.merge(data1, on="secID")

    data2[u"商誉占净资产比例"]=np.round(data2["goodwill"]/(data2.TAssets-data2.TLiab-data2.minorityInt),2)
    data2[u"商誉占净利润比例"]=np.round(data2["goodwill"]/data2.NIncomeAttrP,2)

    # 排序
    data2=data2[(np.abs(data2[u"商誉占净利润比例"])>=1)|(data2[u"商誉占净资产比例"]>=0.3)].sort_values(by=[u"商誉占净资产比例",u"商誉占净利润比例"],ascending=[False,False]).reset_index(drop=True)
    if withPrint:
        print(data2.to_html())
    return data2.secID.tolist()

def goodWillBaoleiTicker(ticker,withPrint=True):
    """
    找出并打印2019年3季报商誉存雷的股票
    eg.
    # goodWillBaolei(set_universe("HS300"))
    """
    thisYear="2019"
    # 合并资产负债表(3季度)
    data = DataAPI.FdmtBSGet(ticker=ticker,reportType=u"Q3",beginYear=thisYear,endYear=thisYear,
                             field=u"ticker,goodwill,TAssets,TLiab,minorityInt",pandas="1")
    data=data.dropna()
    data=data.drop_duplicates()
    # 净利润表（3季度）
    field=u"ticker,NIncomeAttrP,secShortName"
    data1=DataAPI.FdmtISGet(ticker=ticker,reportType=u"CQ3",beginYear=thisYear,endYear=thisYear,field=field,pandas="1")

    data2=data.merge(data1, on="ticker")

    data2[u"商誉占净资产比例"]=np.round(data2["goodwill"]/(data2.TAssets-data2.TLiab-data2.minorityInt),2)
    data2[u"商誉占净利润比例"]=np.round(data2["goodwill"]/data2.NIncomeAttrP,2)

    # 排序
    data2=data2[(np.abs(data2[u"商誉占净利润比例"])>=1)|(data2[u"商誉占净资产比例"]>=0.3)].sort_values(by=[u"商誉占净资产比例",u"商誉占净利润比例"],ascending=[False,False]).reset_index(drop=True)
    if withPrint:
        print(data2.to_html())
    return data2.ticker.tolist()
def goodWillDistByTicker(ticker,withPrint=True):
    """
    商誉暴雷情况分布
    """
    thisYear="2019"
    # 合并资产负债表(3季度)
    data = DataAPI.FdmtBSGet(ticker=ticker,reportType=u"Q3",beginYear=thisYear,endYear=thisYear,
                             field=u"ticker,goodwill,TAssets,TLiab,minorityInt",pandas="1")
    data=data.dropna()
    data=data.drop_duplicates()
    # 净利润表（3季度）
    field=u"ticker,NIncomeAttrP,secShortName"
    data1=DataAPI.FdmtISGet(ticker=ticker,reportType=u"CQ3",beginYear=thisYear,endYear=thisYear,field=field,pandas="1")

    data2=data.merge(data1, on="ticker")

    data2[u"商誉占净资产比例"]=np.round(data2["goodwill"]/(data2.TAssets-data2.TLiab-data2.minorityInt),2)
    data2[u"商誉占净利润比例"]=np.round(data2["goodwill"]/data2.NIncomeAttrP,2)
    data2[u'暴雷风险']=np.where((np.abs(data2[u"商誉占净利润比例"])>=1)|(data2[u"商誉占净资产比例"]>=0.3),1,0)
    # 排序
    data2=data2.sort_values(by=[u'暴雷风险',u"商誉占净资产比例",u"商誉占净利润比例"],ascending=[False,False,False]).reset_index(drop=True)
    if withPrint:
        print(data2[data2[u'暴雷风险']==1].to_html())
    return data2
def extGoodWillTicker(ticker,withPrint=True,symbol_or=True):
    """
    排除商誉暴雷股票后的ticker
    """
    thisYear="2019"
    # 合并资产负债表(3季度)
    data = DataAPI.FdmtBSGet(ticker=ticker,reportType=u"Q3",beginYear=thisYear,endYear=thisYear,
                             field=u"ticker,goodwill,TAssets,TLiab,minorityInt",pandas="1")
    data=data.dropna()
    data=data.drop_duplicates()
    # 净利润表（3季度）
    field=u"ticker,NIncomeAttrP,secShortName"
    data1=DataAPI.FdmtISGet(ticker=ticker,reportType=u"CQ3",beginYear=thisYear,endYear=thisYear,field=field,pandas="1")

    data2=data.merge(data1, on="ticker")

    data2[u"商誉占净资产比例"]=np.round(data2["goodwill"]/(data2.TAssets-data2.TLiab-data2.minorityInt),2)
    data2[u"商誉占净利润比例"]=np.round(data2["goodwill"]/data2.NIncomeAttrP,2)
    
    # 符号 | or &
    if symbol_or:
        data2[u'暴雷风险']=np.where((np.abs(data2[u"商誉占净利润比例"])>=1)|(data2[u"商誉占净资产比例"]>=0.3),1,0)
    else:
        data2[u'暴雷风险']=np.where((np.abs(data2[u"商誉占净利润比例"])>=1)&(data2[u"商誉占净资产比例"]>=0.3),1,0)

    # 排序
    #data2=data2.sort_values(by=[u'暴雷风险',u"商誉占净资产比例",u"商誉占净利润比例"],ascending=[False,False,False]).reset_index(drop=True)
    if withPrint:
        print(r"以下可能存在暴雷风险")
        print(data2[data2[u'暴雷风险']==1].to_html())
    return list(set(ticker)-set(data2[data2[u'暴雷风险']==0].ticker.values))
def ma5_up_ma10_week(secIDs,beginDate=None):
    """
    5周均线上穿10周均线（后复权）
    """
    if beginDate is None:
        beginDate=cal.advanceDate(date.today(), '-60B',  BizDayConvention.Preceding).strftime('%Y%m%d')
    data=DataAPI.MktEquwAdjAfGet(secID=secIDs,beginDate=beginDate,field=u"endDate,secID,closePrice",pandas="1")
    # 通过数据透视转换数据格式
    data = data.pivot_table(index='endDate', columns='secID', values='closePrice')
    data=data.dropna(axis='columns')
    ma5=data.rolling(5).mean().dropna()
    ma10=data.rolling(10).mean().dropna()
    cha=(ma5-ma10).dropna().T
    # cha=cha[(cha.ix[:,-3]<0)&(cha.ix[:,-2]<0)&(cha.ix[:,-1]>0)]
    cha=cha[(cha.ix[:,-2]<0)&(cha.ix[:,-1]>0)]
    return cha.index.tolist()
def ma5_up_ma10_week_ticker(tickers,beginDate=None):
    """
    5周均线上穿10周均线（后复权）
    """
    if beginDate is None:
        beginDate=cal.advanceDate(date.today(), '-60B',  BizDayConvention.Preceding).strftime('%Y%m%d')
    data=DataAPI.MktEquwAdjAfGet(ticker=tickers,beginDate=beginDate,field=u"endDate,ticker,closePrice",pandas="1")
    # 通过数据透视转换数据格式
    data = data.pivot_table(index='endDate', columns='ticker', values='closePrice')
    data=data.dropna(axis='columns')
    ma5=data.rolling(5).mean().dropna()
    ma10=data.rolling(10).mean().dropna()
    cha=(ma5-ma10).dropna().T
    # cha=cha[(cha.ix[:,-3]<0)&(cha.ix[:,-2]<0)&(cha.ix[:,-1]>0)]
    cha=cha[(cha.ix[:,-2]<0)&(cha.ix[:,-1]>0)]
    return cha.index.tolist()
def majorContract(ticker=None, beginDate=None, returnDf=True):
    """
    重大合同
    returnDf 是否返回dataFrame,默认是，否则返回ticker
    """
    if beginDate is None:
        beginDate=myTradeDate('-2B')
    if ticker is None:
        ticker=universeTicker()
    name_dict={
        "updateTime":r"更新时间",
        "reportProgress":r"项目进度",
        "projectName":r"项目名称",
        "contractGrand":r"合同金额上限",
        "biddingAmount":r"合同金额/上年度营业收入（%）",
        "isrelatedTransaction":r"关联交易",
        "announcementId":r"公告ID"}
    # 重大合同
    field="updateTime,ticker,secShortName,reportProgress,projectName,contractGrand,biddingAmount,isrelatedTransaction,announcementId"
    data=DataAPI.MajorContractGet(ticker=ticker,beginDate=beginDate,field=field,pandas="1")
    if returnDf:
        return data.rename(columns=name_dict)
    return list(set(data.ticker.values))

def preAnnualReport(ticker=None,beginDate=None, returnDf=True,):
    """
    查询业绩预告信息
    """
    if beginDate is None:
        beginDate=myTradeDate('-2B')
    if ticker is None:
        ticker=universeTicker()
    data=DataAPI.FdmtEfGet(ticker=ticker,publishDateBegin=beginDate,field=u"",pandas="1")
    if returnDf:
        return data
    return list(set(data.ticker.values))
def preAnnualReportFromReport(ticker=None,beginDate=None, returnDf=True,searchkey=r"业绩预告"):
    """
    业绩预告-数据来自公告
    从公告里面搜索-包含业绩预告的企业
    beginDate 开始时间（N个交易日）
    returnDf 是否返回dataFrame
    """
    if beginDate is None:
        beginDate=myTradeDate('-2B')
    if ticker is None:
        ticker=universeTicker()
    
    # 根据公告ID获取公告内容
    field="tiker,reportID,publishDate,reportType,Category"
    data=DataAPI.ReportByTickerGet(ticker=ticker,beginDate=beginDate,endDate=u"",field=field,pandas="1")
    reportIDs=data.reportID.apply(str).tolist()
    data1=DataAPI.AnnoProfitabilityGet(reportID=reportIDs,beginDate=u"",endDate=u"",
                                       field="reportID,zsAutoCategory,profitablity",pandas="1")
    data2=DataAPI.ReportContentByIDGet(reportID=reportIDs,field=u"reportID,URL",pandas="1")
    data=data.merge(data1, on="reportID", how="left")
    data=data.merge(data2, on="reportID", how="left")
    # data.ix[:, :-1]
    # drop掉Category=nan的行
    data=data.dropna(subset=["Category"],how='any')
    data=data[data["Category"].str.contains(searchkey)]
    if returnDf:
        return data
    return list(set(data.tiker.values))

def preAnnualReport(ticker=None,beginDate=None, returnDf=True,rename=True):
    """
    查询业绩预告信息
    """
    if beginDate is None:
        beginDate=myTradeDate('-2B')
    if ticker is None:
        ticker=universeTicker()
    name_dict={
        "mergedFlag":r"合并类型",
        "revChgrLL":r"收入增减幅度下限(%)",
        "revChgrUPL":r"收入增减幅度上限(%)",
        "expRevLL":r"预期收入下限",
        "expRevUPL":r"预期收入上限",
        "NIncomeChgrLL":r"净利润增减幅度下限(%)",
        "NIncomeChgrUPL":r"净利润增减幅度上限(%)",
        "expnIncomeLL":r"预期净利润下限",
        "expnIncomeUPL":r"预期净利润上限",
        "NIncAPChgrLL":r"归母净利润增减幅度下限(%)",
        "NIncAPChgrUPL":r"归母净利润增减幅度上限(%)",
        "expnIncAPLL":r"预期归母净利润下限",
        "expnIncAPUPL":r"预期归母净利润上限",
        "expEPSLL":r"预期每股收益下限",
        "expEPSUPL":r"预期每股收益上限",
        
    }
    field="ticker,secShortName,endDate,mergedFlag,revChgrLL,\
    revChgrUPL,expRevLL,expRevUPL,NIncomeChgrLL,NIncomeChgrUPL,expnIncomeLL,expnIncomeUPL,\
    NIncAPChgrLL,NIncAPChgrUPL,expnIncAPLL,expnIncAPUPL,expEPSLL,expEPSUPL"
    data=DataAPI.FdmtEfGet(ticker=ticker,publishDateBegin=beginDate,field=field,pandas="1")
    if returnDf:
        if rename:
            return data.rename(columns=name_dict)
        return data
    return list(set(data.ticker.values))

def annualReport(ticker=None,beginDate=None, printIt=False,searchkey=None):
    """
    获取公告
    beginDate 开始时间（N个交易日）
    """
    if beginDate is None:
        beginDate=myTradeDate('-5B')
    if ticker is None:
        ticker=universeTicker()
    
    # 根据公告ID获取公告内容
    field="tiker,reportID,publishDate,reportType,Category"
    data=DataAPI.ReportByTickerGet(ticker=ticker,beginDate=beginDate,endDate=u"",field=field,pandas="1")
    reportIDs=data.reportID.apply(str).tolist()
    data1=DataAPI.AnnoProfitabilityGet(reportID=reportIDs,beginDate=u"",endDate=u"",
                                       field="reportID,zsAutoCategory,profitablity",pandas="1")
    data2=DataAPI.ReportContentByIDGet(reportID=reportIDs,field=u"reportID,URL",pandas="1")
    data=data.merge(data1, on="reportID", how="left")
    data=data.merge(data2, on="reportID", how="left")
    
    if searchkey is not None:
        # drop掉Category=nan的行
        data=data.dropna(subset=["Category"],how='any')
        data=data[data["Category"].str.contains(searchkey)]
    if printIt:
        print(data.to_html())
    return data
def stockInfoByName(namestr, returnDf=False):
    """
    根据名字查看股票基本信息
    """
    namelist=namestr.split()
    # 查看股票基本信息
    field="ticker,ListSector,secShortName,secFullName,listDate,totalShares,nonrestFloatShares,officeAddr,primeOperating"
    data=DataAPI.EquGet(secID=u"",ticker=u"",equTypeCD=u"A",listStatusCD=u"",exchangeCD="",ListSectorCD=u"1,2,3",field=field,pandas="1")
    data=data[data.secShortName.isin(namelist)]

    # 查看二级分类
    ticker=data.ticker.values
    field="ticker,isNew,industryID2,industryName2,industryName3"
    data1=DataAPI.EquIndustryGet(ticker=ticker,industryVersionCD=u"010303",field=field,pandas="1")
    data1=data1[data1["isNew"]==1]
    del data1["isNew"]
    data=data.merge(data1, on='ticker').sort("industryName2")

    # 列的顺序
    cols=["ticker","ListSector","secShortName","secFullName","industryName2","industryName3","listDate","totalShares",
          "nonrestFloatShares","officeAddr","primeOperating","industryID2",]
    print(data[cols].to_html())
    if returnDf:
        return data[cols]
    return data.ticker.values

def zengchiByTicker(ticker,beginDate=None):
    """
    高管增持股票
    """
    if beginDate is None:
        beginDate=myTradeDate('-4B')
    print(u"开始日期=("+beginDate+")")
    # 获取持仓变动情况
    raw_df=DataAPI.EquMsChangesGet(ticker=ticker,
                            beginDate=beginDate,endDate="",fillDateBegin=u"",fillDateEnd=u"",
                            field=u"ticker,changeReason,changeDate,shareChanges,avgPrice",pandas="1")
    raw_df['change_money']=raw_df['shareChanges'] * raw_df['avgPrice']
    raw_df["change_money"]=raw_df['change_money'].fillna(0).astype(int)
    del raw_df['shareChanges']
    del raw_df['avgPrice']
    raw_df=raw_df.groupby(["changeDate","ticker"])["change_money"].sum()
    raw_df=pd.DataFrame(raw_df,columns=["change_money"]).reset_index()

    #raw_df.rename(columns = {"changeDate": "date"},  inplace=True)
    raw_df = raw_df[['change_money','ticker', 'changeDate']]
    raw_df.index = pd.to_datetime(raw_df['changeDate'])

    #获取股票名称
    data=DataAPI.EquGet(secID=u"",ticker=raw_df["ticker"].tolist(),equTypeCD=u"A",listStatusCD=u"",exchangeCD="",field=u"",pandas="1")
    name_dict=data.set_index("ticker")["secShortName"].to_dict()
    raw_df["secShortName"]=raw_df["ticker"].map(name_dict)
    raw_df=raw_df[(raw_df["change_money"]>=3000000) | (raw_df["change_money"]<=-3000000)]

    if len(raw_df):
        raw_df["buy"]=(raw_df["change_money"]).apply(lambda x : 1 if x>=3000000 else 0)
        raw_df["sale"]=(raw_df["change_money"]).apply(lambda x : 1 if x<=-3000000 else 0)
        print(raw_df.to_html())
        tmp=raw_df[["secShortName","buy","sale"]].groupby(by=["secShortName"]).sum()
        print(tmp.sort(["buy","sale"],ascending=[False,False]).to_html())
        return tmp
    return None
def b_breaker(ticker, preTradeDate=None, todayData=None):
    """
    日内交易策略 b-breaker
    newData=(今日最高价，今日当时价格，几日最低价)
    b-breaker日内交易策略
    """
    if preTradeDate is None:
        preTradeDate=myTradeDate('-1B')
    field='ticker,secShortName,highestPrice,lowestPrice,closePrice'
    data=DataAPI.MktEqudGet(ticker=ticker,tradeDate=preTradeDate,isOpen="1",field=field,pandas="1")
    # 观察卖出价 = High+0.35*(Close-Low)
    data['saleSetup']=data['highestPrice']+0.35*(data['closePrice']-data['lowestPrice'])
    # 观察买入价 = Low-0.35*(Low-Close)
    data['buySetup']=data['lowestPrice']-0.35*(data['lowestPrice']-data['closePrice'])
    # 反转卖出价 = 1.07/2*(High+Low)-0.07*Low
    data['saleEnter']=1.07/2*(data['highestPrice']+data['lowestPrice'])-0.07*data['lowestPrice']
    # 反转买入价 = 1.07/2*(High+Low)-0.07*High
    data['buyEnter']=1.07/2*(data['highestPrice']+data['lowestPrice'])-0.07*data['highestPrice']
    # 突破买入价 = 观察卖出价+0.25*(观察卖出价-观察买入价)
    data['buyBreak']=data['saleSetup']+0.25*(data['saleSetup']-data['buySetup'])
    # 突破卖出价 = 观察买入价+0.25*(观察卖出价-观察买入价)
    data['saleBreak']=data['buySetup']+0.25*(data['saleSetup']-data['buySetup'])
    
    data['high']=data['highestPrice']
    data['price']=data['closePrice']
    data['low']=data['lowestPrice']
    
    if todayData is not None:
        data=data.merge(todayData, on='ticker')
    
    data['qushi']=np.where(data['price']-data['buyBreak']>0,1,0)
    data['qushi']=np.where(data['price']-data['buyBreak']<0,-1,data['qushi'])
    
    def cond(x):
        if (x.high-x.saleSetup>0) and (x.price-x.saleEnter<0):
            return -1
        if x.low-x.buySetup<0 and x.price-x.buyEnter>0:
            return 1
        return 0
    data['fanzhuan']=data.apply(cond,axis=1)
    # 日内采用趋势策略
    # if today_now >= data['buyBreak'].values[0]:
    #     print("当前价格>="+str(data['buyBreak'].values[0])+"-趋势策略-突破买入")
    # if today_now <= data['saleBreak'].values[0]:
    #     print("当前价格<="+str(data['saleBreak'].values[0])+"-趋势策略-突破卖出")
    return data
```
