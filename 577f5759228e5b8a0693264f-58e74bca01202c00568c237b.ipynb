**3，交易策略**
   根据上文所述，尝试设计一个常规的底部十字星策略。我们将资金等额分成20份，依照参数（h=0.1%,a=3,b=3）在沪深300股票池中识别底部十字星信号，每日依次动用这二十份资金中的一份等量购买出现信号的股票，持有50天后出售，出售获得的资金进行下一轮投资，各份资金保持独立。

   回测显示，这一策略的市场表现还可以，但似乎并没有完全体现出十字星本身的优势，[1]中采用的是改进方法是超额十字星策略，这似乎是一个研究的方向，等通联数据给出指数的分钟数据后可以进行进一步的分析。
