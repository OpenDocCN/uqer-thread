# VWAP策略

此策略来自于UQER的[帮助文档](https://uqer.datayes.com/help/faq/#VWAP策略)
* **策略思路**

计算过去一段时间内的平均成交价（VWAP）。
当昨日最低价低于VWAP的某一阈值时，买入；当昨日最低价高于VWAP时，卖出。
***
这一策略初听起来很适合日内做T，但是再想一想这个策略不就是在拥抱熊市，躲过牛市吗？
研究了Uqer的几个策略实例，几个策略都挺好的，但都只是从单方面举例说明，真正实际应用还是要融合一些其他的策略。不过对于[万圣节效用](https://uqer.datayes.com/v3/community/share/5ccddcd3a7a3a2012a1add28)，这个策略我挺喜欢的，简单有效，对于普通小散户还是很不错的策略。

**在原策略上进行小的升级，避开熊市，比原策略好，但还是没有跑过指数，因为策略没有拥抱牛市，这个策略思路比较适合震荡市**

改变下，拥抱牛市，目前对于牛熊市的判断过于简单，后面会详细的研究牛熊市的特征

深入的研究了几个策略，一个完整的策略应该达到的功能：
--
* 在牛市，选择高风险对应的高收益策略，拥抱牛市
* 震荡市，选择有效的震荡市策略，像这篇里面VWAP策略就能应付震荡市，当然应该还有更好的策略
* 回避熊市

对于以上3点，关键在于识别牛熊市，牛熊市之外的属于震荡市。对于牛熊市的特征识别，后续将引用深度学习方面的知识进行一番研究。
