# LCAP减LFLO

不知道优矿有没这个因子，跑来看看，还是有点用的。17年依然崩。

```python
def rClist(pday,qin,SorL):
    ratioC = DataAPI.MktStockFactorsOneDayGet(tradeDate=pday,secID=u"",field=u"secID,LCAP,LFLO",pandas="1").set_index('secID')
    ratioC = ratioC.dropna()
    ratioC['rC'] = ratioC['LCAP']-ratioC['LFLO']
    ratioC = ratioC[['rC']]
    ratioC = ratioC[ratioC['rC'] !=0]
    qq = ratioC['rC'].quantile(q=qin)
    if SorL == 'larger':
        ratioC = ratioC[ratioC['rC'] >qq]
    elif SorL == 'smaller':
        ratioC = ratioC[ratioC['rC'] <qq]
    
    return ratioC.index.tolist()
```
