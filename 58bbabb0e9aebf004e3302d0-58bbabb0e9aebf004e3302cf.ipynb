```python
import datetime
import math
import pandas as pd
import numpy as np
from CAL.PyCAL import *
cal = Calendar('China.SSE')
font.set_size(15)
```

```python
todaydate=datetime.datetime.now().strftime('%Y-%m-%d')
lastTD=cal.advanceDate(todaydate,'-1B',BizDayConvention.Preceding).strftime('%Y-%m-%d')#last trading date
```

```python
#新股
security_listDate=DataAPI.EquGet(equTypeCD=u"A",secID=u"",ticker=u"",listStatusCD=u"",field=u"exchangeCD,equType,ListSectorCD,ListSector,secID,ticker,secShortName,ListDate,officeAddr,primeOperating,totalShares,nonrestFloatShares,nonrestfloatA,TShEquity",pandas="1")
new_equity = security_listDate[security_listDate['listDate'] > '2015-01-01'].reset_index(drop='False')
new_equity = new_equity[new_equity['listDate'] < todaydate].reset_index(drop='False')

#+行业
ind=DataAPI.EquIndustryGet(industryVersionCD=u"010303",industry=u"",secID=u"",ticker=new_equity['ticker'],intoDate=u"",field=u"isNew,ticker,industryName1,industryName2,industryName3",pandas="1")
ind=ind.loc[ind['isNew']==1]

new_equity=pd.merge(new_equity , ind,how='left',on='ticker')

#+地域
addr=DataAPI.SecTypeRegionRelGet(typeID=u"",ticker=new_equity['ticker'],field=u"ticker,typeName",pandas="1")
new_equity=pd.merge(new_equity , addr,how='left',on='ticker')


#新股连板天数
Cont=DataAPI.MktIpoConTraddaysGet(ticker=new_equity['ticker']
                                                      ,field=u"ticker,continiousTradDays,breakConClosePrice,isContinue"
                                                      ,pandas="1")
new_equity=pd.merge(new_equity , Cont,how='left',on='ticker')


#新股破板后最高收盘价

new_equity['listDate'] = pd.to_datetime(new_equity['listDate'],format="%Y-%m-%d")
b=[]
for j in range(new_equity.shape[0]):
    if new_equity['isContinue'][j]:
        b.append(lastTD)
    else:
        following='+'+str(new_equity['continiousTradDays'][j])+'B'
        b.append(cal.advanceDate(new_equity['listDate'][j],following,BizDayConvention.Following).strftime('%Y-%m-%d'))
        
new_equity['breakDate'] = pd.Series(b,index=new_equity.index)

h=[]
for j in range(new_equity.shape[0]):
    datah=DataAPI.MktEqudAdjGet(secID=u"",ticker=new_equity['ticker'][j]
                      ,isOpen=u"1",beginDate=u'',endDate=todaydate,field=u"closePrice",pandas="1")
    if new_equity['continiousTradDays'][j] == len(datah['closePrice']):
        h.append(max(datah['closePrice'][(new_equity['continiousTradDays'][j]-1):]))
    else:
        h.append(max(datah['closePrice'][(new_equity['continiousTradDays'][j]):]))
    
new_equity['HightestAdCl'] = pd.Series(h,index=new_equity.index)
```
