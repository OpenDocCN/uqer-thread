# Python 相关报错

## TypeError: 'NoneType' object has no attribute '__getitem__'

```python
TypeError: 'NoneType' object has no attribute '__getitem__'
```

## 解决办法

 1. 观察是否存在变量为空的情况，比如持仓为空。
 2. 添加异常处理逻辑，进行异常处理。
 3. 或者完善相关代码避免出现异常情况，见下面示例。

## 示例

下面的例子，使用了 stock_account.get_position(stock) 获取某个股票的持仓信息，但当前并没有持有该股票。
这种情况会报错。

```python
accounts = {
    'stock_account': AccountConfig(account_type='security', capital_base=10000000)
}

def handle_data(context):    
    current_universe = context.get_universe(exclude_halt=True)
    for stock in current_universe:
        stock_account = context.get_account('stock_account')  
        amount = stock_account.get_position(stock)['amount']
```

修改为下面的写法，就可以避免这个问题。

```python
accounts = {
    'stock_account': AccountConfig(account_type='security', capital_base=10000000)
}

def handle_data(context):    
    current_universe = context.get_universe(exclude_halt=True)
    for stock in current_universe:
        stock_account = context.get_account('stock_account')  
        try:
            amount = stock_account.get_position(stock)['amount']
        except:
            amount = 0
```

修改为下面的写法，也可以避免这个问题。

```python
accounts = {
    'stock_account': AccountConfig(account_type='security', capital_base=10000000)
}

def handle_data(context):    
    current_universe = context.get_universe(exclude_halt=True)
    for stock in current_universe:
        stock_account = context.get_account('stock_account')  
        current_position = stock_account.get_position(stock)
        if current_position:
            amount = current_position.amount
        else:
            amount = 0
```
