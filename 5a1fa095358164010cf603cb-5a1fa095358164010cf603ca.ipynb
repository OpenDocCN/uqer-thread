# 交易的胜率、赔率——如何在交易中获取优势

伯努利分布是很简单的离散分布， 在伯努利分布中，随机变量只有两个可能的取值：即1和0.如果随机变量取值1的概率为P， 那么随机变量取值为0的概率就为1-P


在Numpy中使用 numpy.random.binomial（1， p）来获取1的概率为p的前提下生成的随机变量。如果p=0.5的话， 那么就是类似投掷硬币的结果， 即正面在上和反面在上的概率相同
下面使用binomial来做一个综合实例。

## 如何在交易中获取优势

在交易中，交易者永远是处于不利的地位的， 不利的情况就是需要交手续费， 你可能会说那个没有很多钱（很多交易者， 特别是在牛市中挣到钱的人都是这种想法），如果你随机地胡乱买卖股票， 在有足够多的本钱的情况下交易足够多的次数的话，最后的结果一定是符合正态分布的；但是如果每次要多交1%的手续费的话，情况就大不一样了，赌场要抽头， 不抽头的赌场一定有“老千”。

自定义函数casino（）：假设有100个赌徒， 每个赌徒都有100000元， 并且每个人都想在赌场里面玩100万次， 在不同的胜率、赔率和手续费下 casino（）函数返回最终的统计结果

```python
import numpy as np
import  matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("dark")
```

```python
# 设置100个赌徒
gamblers = 100
def casino(win_rate, win_once=1, loss_once=1, commission=0.01):
    """
    赌场：简单设定每个赌徒都有100000元， 并且每个赌徒都想在赌场里面玩1000000次，
    但是如果没钱了就被想玩了
    win_rate： 赢的概率
    win_once： 每次赢的钱数
    loss_once： 每次输的钱数
    commission：手续费在这里简单设置为0.01 即 1%
    """
    my_money = 100000
    play_cnt = 1000000
    commission = commission
    for _ in np.arange(0, play_cnt):
        # 使用伯努利分布， 根据win_rate 来获取输赢
        w = np.random.binomial(1, win_rate)
        if w:
            # 赢了 + win_once
            my_money += win_once
        else:
            # 输了 - loss_once
            my_money -= loss_once
        # 每玩一把的手续费
        my_money -= commission
        if my_money <= 0:
            # 没钱就没得玩了， 不赊账
            break
    return my_money
```

* 假设天堂赌场， 没有“老千”， 也没有抽头的情况(也就是说赌徒的胜率=0.5， 赔率=1， 手续费=0)

```python
heaven_casino = [casino(win_rate=0.5, win_once=1, loss_once=1, commission=0) for _ in np.arange(0, gamblers)]
```

```python
fig = plt.figure(figsize=(10, 6))
ax_heaven = fig.add_subplot(1,1,1)
ax_heaven = plt.hist(heaven_casino, bins=30)
plt.grid()
print ("天堂赌场100个赌徒最后的平均资产为：{}".format(np.mean(heaven_casino)))
print ("天堂赌场100个赌徒的最后资产的标准差为：{:.2f}".format(np.std(heaven_casino)))
print ("天堂赌场100个赌徒最后最少资产为：{}".format(np.min(heaven_casino)))
print ("天堂赌场100个赌徒最后最多资产为：{}".format(np.max(heaven_casino)))
```

可以看到，在天堂赌场中，所有人都活下来了，玩了100万次， 每个人亏和赚的最后都不太多， 赚的最多的是2282元， 亏得最多的是1816元。

*  假设老千赌场， 赌场出老千， 没有抽头， 赌徒的胜率只有40%的情况下(也就是说赌徒的胜率=0.4， 赔率=1， 手续费=0)

```python
LaoQian_casino = [casino(win_rate=0.4, win_once=1, loss_once=1, commission=0) for _ in np.arange(0, gamblers)]
```

```python
fig = plt.figure(figsize=(10, 6))
ax_heaven = fig.add_subplot(1,1,1)
ax_heaven = plt.hist(LaoQian_casino, bins=30)
plt.grid()
print ("老千赌场100个赌徒最后的平均资产为：{}".format(np.mean(LaoQian_casino)))
print ("老千赌场100个赌徒的最后资产的标准差为：{:.2f}".format(np.std(LaoQian_casino)))
print ("老千赌场100个赌徒最后最少资产为：{}".format(np.min(LaoQian_casino)))
print ("老千赌场100个赌徒最后最多资产为：{}".format(np.max(LaoQian_casino)))
```

结果可以看到，有老千的赌场没有人能够活下来， 钱都归零了，很惨。

* 假设普通赌场， 没有老千存在， 但是有抽头， 每玩一次手续费0.01的情况下（也就是说赌徒的胜率=0.5， 赔率=1， 手续费=0.01）

```python
common_casino = [casino(win_rate=0.5, win_once=1, loss_once=1, commission=0.01) for _ in np.arange(0, gamblers)]
```

```python
fig = plt.figure(figsize=(10, 6))
ax_heaven = fig.add_subplot(1,1,1)
ax_heaven = plt.hist(common_casino, bins=30)
plt.grid()
print ("普通赌场100个赌徒最后的平均资产为：{:.2f}".format(np.mean(common_casino)))
print ("普通赌场100个赌徒的最后资产的标准差为：{:.2f}".format(np.std(common_casino)))
print ("普通赌场100个赌徒最后最少资产为：{:.2f}".format(np.min(common_casino)))
print ("普通赌场100个赌徒最后最多资产为：{:.2f}".format(np.max(common_casino)))
```

可以看到有抽头的赌场， 没有一个人是赚钱的， 都是亏钱的， 同时如果当玩的次数再大一个数量级的时候， 可以预见的结果是， 最后所有人的结果肯定也都是0.

现在假设我们的投资市场也是没有“老千”的， 但是一定会有手续费， 这样的话注意上面的赌场函数， 我们不能期待自己有“老千”的能力， 也就是说如果你能有内部消息， 胜率就会提高很多。到60%，或者更高！
那我们剩下来的就只有每次赢钱和书签的参数了。下面把参数设置为每次赢相比默认值win_once多出两个点， 即提高win_once=1.0 -》1.02， 每次输相比默认值loss_once少两个点 ，即降低 loss_once=1.0 -》 0.98， 这样的话赔率即为 1.02/0.98 = 1.040816， 下面让100个赌徒进入这个非均衡赔率的赌场.

```python
inbalance_casino = [casino(win_rate=0.5, win_once=1.02, loss_once=0.98, commission=0.01) for _ in np.arange(0, gamblers)]
```

```python
fig = plt.figure(figsize=(10, 6))
ax_heaven = fig.add_subplot(1,1,1)
ax_heaven = plt.hist(inbalance_casino, bins=30)
plt.grid()
print ("非均衡赔率赌场100个赌徒最后的平均资产为：{:.2f}".format(np.mean(inbalance_casino)))
print ("非均衡赔率赌场100个赌徒的最后资产的标准差为：{:.2f}".format(np.std(inbalance_casino)))
print ("非均衡赔率赌场100个赌徒最后最少资产为：{:.2f}".format(np.min(inbalance_casino)))
print ("非均衡赔率赌场100个赌徒最后最多资产为：{:.2f}".format(np.max(inbalance_casino)))
```

从上面的结果中我们可以看到， 每个人最终都能从赌场中赢到钱。

不过这也是不正常的，凭什么赌徒能做到每次赢的时候多赢，输的时候少输呢？

事实上每本关于交易的书籍都会提到止损的策略， 这些策略的核心思想就是让赚钱的股票多赚钱， 让亏损的股票少亏损（让盈利奔跑， 让亏损尽快了结）。这样的结果就是亏损的次数多， 每次赢的钱比每次亏的钱多， 也就是高止盈位， 低止损位。
使用上述的止盈止损策略后， 会降低胜率， 所以不能平白无故低提高win_once， 对应的要降低 win_rate。这里假设win_rate是0.45，即胜率为45%， 赔率依然使用上面的设置 win_once = 1.02， loss_once = 0.98，即赔率依然是1.041， 下面让一百个赌徒进入这个非均衡胜率和非均衡赔率共同生效的赌场。

```python
inbalance2_casino = [casino(win_rate=0.45, win_once=1.02, loss_once=0.98, commission=0.01) for _ in np.arange(0, gamblers)]
```

```python
fig = plt.figure(figsize=(10, 6))
ax_heaven = fig.add_subplot(1,1,1)
ax_heaven = plt.hist(inbalance2_casino, bins=30)
plt.grid()
print ("非均衡胜率和赔率赌场100个赌徒最后的平均资产为：{:.2f}".format(np.mean(inbalance2_casino)))
print ("非均衡胜率和赔率赌场100个赌徒的最后资产的标准差为：{:.2f}".format(np.std(inbalance2_casino)))
print ("非均衡胜率和赔率赌场100个赌徒最后最少资产为：{:.2f}".format(np.min(inbalance2_casino)))
print ("非均衡胜率和赔率赌场100个赌徒最后最多资产为：{:.2f}".format(np.max(inbalance2_casino)))
```

可以看到， 最后的结果就是又回到了有输有赢的情况了，也就是产生了天堂赌场的感觉。

现实中也可以看到多数交易者的胜率非常高， 但是他们的账户最终都是亏损的， 交易中最虚幻的就是胜率， 但是大多数人追求的反而是胜率， 不关心盈亏比等其他更重要的因素， **量化交易的核心还是交易， 交易的基本法则依然适用！**
