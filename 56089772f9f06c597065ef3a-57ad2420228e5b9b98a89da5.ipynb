```python
# 第一步获取策略所需要的数据
from datetime import datetime
start = datetime(2015, 1, 1)
end   = datetime(2015, 9, 5)
benchmark = 'HS300'
universe  = set_universe('SH50')
capital_base = 100000

sim_params = quartz.sim_condition.env.SimulationParameters(start, end, benchmark, universe, capital_base)
idxmap_all, data_all = quartz.sim_condition.data_generator.get_daily_data(sim_params)

original_universe = universe
```

```python
# 第二步，编写策略
def initialize(account):
    pass

def handle_data(account):
    
    for stock in universe:
        order(stock, 100)

# 第三步，运行 
strategy = quartz.sim_condition.strategy.TradingStrategy(initialize, handle_data)
perfs = []
bts = []

# 上证50成分股，每10个一组，共分成5组。
for i in xrange(0, 50, 10):
    universe = original_universe[i: i+10]
    print universe
    bt, acct = quartz.quick_backtest(sim_params, strategy, idxmap_all, data_all)
    perf = quartz.perf_parse(bt, acct)
    #记录每个策略的性能指标变量：perf到perf列表，在下个cell会用到。
    perfs.append(perf)
    bts.append(bt)
```

```python
import seaborn
from matplotlib import pylab
from CAL.PyCAL import *
import pandas as pd

# 把每个策略的perf集中到dataframe里，用于画图
df = pd.DataFrame({})
for index, perf in enumerate(perfs):
    df['index%s' %index] = perf['cumulative_values']

pylab.figure(figsize = (16,8))
tradeDate = df.index
# 画出每个策略的性能
for column in df.columns:
    pylab.plot(tradeDate,df[column])

pylab.legend([u'组合%s'%(i+1) for i in xrange(len(df))], prop = font)
pylab.title(u'不同组合的净值比较', fontproperties = font)
```
