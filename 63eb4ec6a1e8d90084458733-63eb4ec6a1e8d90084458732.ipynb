## 2. 数据准备
---
发现优矿的复权的基金行情要专业版的，但这难不倒我们，目前发现DataAPI.FundNavGrGet还能使用，而且每天的收益率也是经过复权的，这个API提供的数据算动量比较麻烦，主要数据量有点大，内存不够用，如果算区间的净值增长的话，我们其实只要开始和结束两个复权净值就可以了。我们分析的时间段为2012年末到最新。

```python
import pandas as pd
import matplotlib.pyplot as plt
from CAL.PyCAL import *
from datetime import datetime, timedelta

import seaborn as sns
sns.set_style('whitegrid')
```

```python
# 获取基金信息
df = DataAPI.FundGet(category=u"E,H",idxID=u"",idxTicker=u"",operationMode=u"O",beginDate="",endDate="",status="",field=u"",pandas="1")

# 选择A类
df = df[(df['isQdii'] == 0) & (df['isFof'] == 0) & (pd.isnull(df['indexFund'])) & (pd.isnull(df['etfLof']))].sort_values(by='secShortName')
df['name'] = df['secShortName'].apply(lambda x: x.split('-')[0])
df = df.drop_duplicates(subset=['name'], keep='first')

df = df[~((df['name'].str.contains('持有')) | ((df['name'].str.contains('LOF')))) ]
```

```python
def get_fund(date, all_fund_info):
    """
    给定日期，并按照所有的基金信息，筛选成立至少满1年的普通股票型，偏股混合型和灵活配置型的基金
    """
    pre_1y = str(int(date[:4]) - 1) + '-' + date[4:6] + '-' + date[6:8]
    current_fund = (all_fund_info[(all_fund_info['establishDate'] <= pre_1y) & (pd.isnull(all_fund_info['expireDate']) | (all_fund_info['expireDate'] > date))]).copy()
    
    return current_fund

def get_report_date_list(date, lag):
    pre_date = pd.to_datetime(date) - timedelta(31*lag * 3)
    date_list = pd.date_range(pre_date, date, freq='m').strftime('%Y%m%d')
    
    date_list = [i for i in date_list if i[4:6] in ('03', '06', '09', '12')]
    
    return date_list[-lag:]

def get_available_fund_info(fund_info, fund_assets):
    r1 = fund_assets['r1'][fund_assets['r1']['equityRatioInTa'] >= 70]['secID']
    r2 = fund_assets['r2'][fund_assets['r2']['equityRatioInTa'] >= 70]['secID']
    r3 = fund_assets['r3'][(fund_assets['r3']['equityRatioInTa'] >= 70) & ((fund_assets['r3']['netAsset'] >= 1e8))]['secID']
    
    fund_list = pd.Index(r1).intersection(pd.Index(r2)).intersection(r3).tolist()
    
    return (fund_info[fund_info['secID'].isin(fund_list)]).copy()
```

```python
ana_date_list = pd.date_range('20121201', '20230131', freq='m').strftime('%Y%m%d')
ana_date_list = [i for i in ana_date_list if i[4:6] in ('01', '04', '07', '10')]

fund_mom_6m = dict()

for date in ana_date_list:
    print(date)
    pre_date = (pd.to_datetime(date) - timedelta(180)).strftime('%Y%m%d')
    current_fund_info = get_fund(date, df)
    date_list = get_report_date_list(date, 3)

    fund_assets = dict()
    fund_assets = {
        'r1': DataAPI.FundAssetsGet(secID=current_fund_info['secID'].tolist(), reportDate=date_list[0], field=u"secID,netAsset,equityRatioInTa",pandas="1"),
        'r2': DataAPI.FundAssetsGet(secID=current_fund_info['secID'].tolist(), reportDate=date_list[1], field=u"secID,netAsset,equityRatioInTa",pandas="1"),
        'r3': DataAPI.FundAssetsGet(secID=current_fund_info['secID'].tolist(), reportDate=date_list[2], field=u"secID,netAsset,equityRatioInTa",pandas="1") 
    }
    
    fund_list = get_available_fund_info(current_fund_info, fund_assets)['secID'].tolist()
    
    fund_nav = DataAPI.FundNavGrGet(secID=fund_list, beginDate=pre_date, endDate=date, field=u"secID,endDate,returnRate",pandas="1")
    fund_nav['returnRate'] = fund_nav['returnRate'] / 100
    fund_nav = fund_nav.pivot(index='endDate', columns='secID', values='returnRate')
    
    fund_mom_6m[date] = (1 + fund_nav).cumprod().iloc[-1, :] - 1
```

与股票不一样，基金如果持有时间过短，交易的费率是很高的，因此我们调仓周期设为一个季度。接下来，我们看看假设持有一个季度的话，这个因子的分组表现怎样。

```python
all_group_return = dict()
for i, date in enumerate(ana_date_list[:-1]):
    print(date)
    all_group_return[ana_date_list[i+1]] = dict()
    
    factor = fund_mom_6m[date].dropna()
    fund_nav = DataAPI.FundNavGrGet(secID=factor.index.tolist(), beginDate=date, endDate=ana_date_list[i+1], field=u"secID,endDate,returnRate",pandas="1")
    fund_nav['returnRate'] = fund_nav['returnRate'] / 100
    fund_nav = fund_nav.pivot(index='endDate', columns='secID', values='returnRate')
    
    fund_nav = (1 + fund_nav).cumprod().iloc[-1, :] - 1
    
    for t in range(10):
        current_group = factor[(factor >= factor.quantile(t/10.)) & (factor <= factor.quantile((t+1)/10.))]
        all_group_return[ana_date_list[i+1]]['group {}'.format(t + 1)] = fund_nav.reindex(current_group.index).mean()
```

```python
print('分组相对平均表现')
ax = all_group_return.sub(all_group_return.mean(axis=1), axis=0).sum().sort_values().plot(kind='bar')
```

可以看到，分组单调性非常好，说明过去6个月基金的净值增长有着不错的区分度。再来看下多空的表现：

```python
print('多空组合')
ax = (1 + all_group_return['group 10'] - all_group_return['group 1']).cumprod().plot()
```

## 3. 小结
* 这个策略在2021年前看着都蛮有效的，但是2021年以后，那些前期涨的多的基金，如果持有一个季度的话明显跑不赢前期涨的少的，说明21年以来，市场的轮动节奏不断的加快，特别是行业上的轮动，这对动量策略来说是个很大挑战。个人感觉基金动量和行业动量有着不小的关联，我们研究下来发现21年4季度以来，传统的量化行业轮动也有很大回撤，在这个的表现上，两者也是比较吻合的。
* 发现对于灵活配置型的基金，经常出现这期入池，下期出池的情况，可能是我对仓位的要求太严格了。
