《主动投资管理组合》中提到，通常因子比历史收益数据更稳定。弄个闹着玩的东西，看看短期内的收益率是不是很稳定。

```python
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
def profit(dt): # 参数为YYYYMMDD的日期
    # 找股票，条件：要中小创吗？反正不要新股和次新股，只要A股非ST
    req_secID = DataAPI.EquGet(equTypeCD=u"A",secID=u"",ticker=u"",listStatusCD=u"L",field=u"",pandas="1")
    req_secID['listDate'] = pd.to_datetime(req_secID['listDate'])
    req_secID = req_secID[req_secID['listDate'] < '2015-05-01'] # 非新股，就粗暴的给了个日期
    # req_secID = req_secID[(req_secID['secID'].str.match('000')) | (req_secID['secID'].str.match('6'))] # 是否要剔除中小创
    
    # field只是表示我试了好多，有些收益归因的时候会是上帝视角……
    df = DataAPI.MktStockFactorsOneDayGet(tradeDate=dt,secID="",ticker=u"",
                                          field=u"ACD20,VOL20,HSIGMA,secID,CurrentRatio,ARTDays,ILLIQUIDITY,EPS,MTM,PS,QuickRatio,FEARNG,LCAP,HBETA,ROE,LFLO,PB,PE,EBITToTOR,ROA,EquityToAsset,Volatility,TOBT",pandas="1")
    ST = DataAPI.SecSTGet(beginDate=dt,endDate=dt,secID=u"",ticker=u"",field=u"secID",pandas="1")
    df = df[~df['secID'].isin(ST['secID'])] # 非ST
    df = df[df['secID'].isin(req_secID['secID'])]
    # df = df[(df['PB'] < 150) & (df['PB']> 0.7)]
    
    # 据说行业因子是结构化风险中最重要的因子
    ind_classification = DataAPI.EquIndustryGet(industryVersionCD=u"010303",industry=u"",secID=u"",ticker=u"",intoDate=u"",
                                                field=u"secID,outDate,industryName1,industryName2",pandas="1")
    ind_classification = ind_classification.loc[ind_classification['outDate'].isnull()] # 只要最新的分类，不要过去它在哪个行业
    res = pd.merge(df,ind_classification,on='secID',how='inner')
    del res['outDate'] # 强迫症发作
    
    # 对月度收益率进行分解，估计因子载荷和因子收益率
    CMRA = DataAPI.MktEqumAdjGet(secID=u"",ticker=u"",monthEndDate=dt,isOpen=u"",beginDate=u"",endDate=u"",field=u"",pandas="1")
    CMRA['CMRA'] = CMRA['chgPct'] # 直接用chgPct表示月度累计收益
    CMRA = CMRA[['secID','turnoverRate','CMRA']] # turnoverRate文档写的是周换手率。。。文档错了吧
    res.dropna(inplace=True)
    res =  pd.merge(res,CMRA,on='secID',how='inner').copy()
    # 返回一个月度收益率，及各种因子数据的大DataFrame
    return res

r_pre = profit(dt = '20160531')
# 缩小下那个DataFrame，顺便弄个deep copy，因为不知道python到底弄了只鸭子（copy）还是弄了只橡皮鸭（view）。
n_pre = r_pre[['LCAP','LFLO','HBETA','CMRA','PB','ARTDays','ACD20','PE','EBITToTOR','ROA','VOL20','HSIGMA','CurrentRatio','FEARNG','QuickRatio',
               'ILLIQUIDITY','PS','turnoverRate','EquityToAsset','Volatility','TOBT','ROE','MTM','EPS']].copy()

# 假设正态分布的话，三倍外是极端不正常的，剔除。
n_pre = n_pre[(n_pre['CMRA'] < n_pre['CMRA'].mean() + 3 * n_pre['CMRA'].std()) & (n_pre['CMRA'] > n_pre['CMRA'].mean() - 3 * n_pre['CMRA'].std())] # 剔除不正常的收益率
# 标准化
n_pre = (n_pre - n_pre.mean()) / n_pre.std()

# 为了上面的标准化，不能在一开始就把str格式的数据放进去，要不然要报错，所以多了如下两句
n_pre['ind'] = r_pre['industryName1']
n_pre['secID'] = r_pre['secID']
n_pre.dropna(inplace=True)

# 多远线性回归……-1表示不要截距，行业才是截距……
ft = smf.ols('CMRA ~ C(ind) + LCAP + HBETA + PB + EBITToTOR + ROA + QuickRatio + turnoverRate +EquityToAsset - 1',n_pre).fit()
print(ft.summary())

# 同上再来一遍
r_now = profit(dt = '20160630')
n_now = r_now[['LCAP','LFLO','HBETA','CMRA','PB','ARTDays','ACD20','PE','EBITToTOR','ROA','VOL20','HSIGMA','CurrentRatio','FEARNG','QuickRatio',
               'ILLIQUIDITY','PS','turnoverRate','EquityToAsset','Volatility','TOBT','ROE','MTM','EPS']].copy()
# 标准化
n_now = (n_now - n_now.mean()) / n_now.std()
n_now['ind'] = r_now['industryName1']
n_now['secID'] = r_now['secID']
n_now.dropna(inplace=True)

# ft.summary()和ft1.summary()的对比，表示，短期内因子其实也不稳定啊。看看行业因子就知道了。连符号都大量的变了，证明资金在调整……
ft1 = smf.ols('CMRA ~ C(ind) + LCAP + HBETA + PB + EBITToTOR + ROA + QuickRatio + turnoverRate + EquityToAsset - 1',n_now).fit()
print(ft1.summary())

fig = plt.figure(figsize=(8,16))
ax1 = fig.add_subplot(211)
# 看看残差分布
ax1.scatter([i for i in range(len(ft.resid))],ft.resid)
ax2 = fig.add_subplot(212)

# 根据5月份的线性回归的系数，来估计六月份的收益，假设六月份的所有其他因子已知，就想求个六月累计收益……
predict = ft.predict(n_now)
# 这是为了保证sec_guess变成DataFrame而不是Series，若n_now['secID']，就需要显示转化成DataFrame。
sec_guess = n_now[['secID']]
sec_guess['predict'] = predict
final = pd.merge(sec_guess,n_now,on='secID',how='inner')

# 意料之中伤感的画了个图……看了图更伤感
ax2.scatter(final['predict'],final['CMRA'])
ax2.set_xlabel('predict')
ax2.set_ylabel('real')
ax2.set_xlim((-3,3))
ax2.set_ylim((-3,3))
fig.show()

```
