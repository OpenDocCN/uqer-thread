## 1.实盘开始时间

```python
live_start_date='2017-01-16'
```

## 2.飞扬一号A

```python
bt_A = bt.copy()
```

```python
A = bt_A.copy()
A = A.rename(columns={'tradeDate':'Date'})
A["Date"] = pd.to_datetime(A["Date"])
A = A.set_index("Date")
A["returns"] = A.loc[live_start_date:,['portfolio_value']].pct_change().fillna(0.)
A = A.returns.dropna()
A.name = u'飞扬一号A'
```

## 3.日度收益率

```python
from lib.data_loader import  get_return,rong_zi_rong_quan

account = pd.DataFrame()
#指数数据
account[u'沪深300'] = get_return(ticker='000300.ZICN', start=live_start_date) # 000016.ZICN 上证50；000300.ZICN 沪深300；399905.ZICN 中证500
account[u'中证500'] = get_return(ticker='399905.ZICN', start=live_start_date) # 000016.ZICN 上证50；000300.ZICN 沪深300；399905.ZICN 中证500


#中投实盘
account3 = pd.read_csv('zt_account.csv',encoding = 'GB2312')
account3['Date'] =  pd.to_datetime(account3['Date']) 
account3 = account3.set_index("Date")
account3 = account3.fillna(0.)
account[u'中投实盘'] = (account3.living_portfolio_value + account3.living_cash + account3.living_dividend).pct_change().fillna(0.)

#财富实盘
account2 = pd.read_csv('cf_account.csv',encoding = 'GB2312')
account2['Date'] =  pd.to_datetime(account2['Date']) 
account2 = account2.set_index("Date")
account2 = account2.fillna(0.)
account[u'财富实盘'] = (account2.living_portfolio_value + account2.living_cash + account2.living_dividend).pct_change().fillna(0.)

#机器模拟
account1 = pd.read_csv('auto_account.csv',encoding = 'GB2312')
account1['Date'] =  pd.to_datetime(account1['Date']) 
account1 = account1.set_index("Date")
account1 = account1.fillna(0.)
account[u'机器模拟'] = (account1.auto_portfolio_value + account1.auto_cash + account1.auto_dividend).pct_change().fillna(0.)

#理论数据
account = account.join(A)



account.tail(5)
```

## 4.累计净值

```python
from matplotlib import pylab

cum_all = (1+account[[u'财富实盘',u'中投实盘',u'机器模拟',u'飞扬一号A',u'沪深300',u'中证500']]).cumprod()
print  '最新净值:\n' ,cum_all.tail(1).to_html()
cum_all.plot(figsize=(16, 9))
pylab.title(u'累计净值走势图',fontproperties=font,fontsize=16)
pylab.legend([u'财富实盘',u'中投实盘',u'机器模拟',u'飞扬一号A',u'沪深300',u'中证500'], prop=font,loc=2);
```
