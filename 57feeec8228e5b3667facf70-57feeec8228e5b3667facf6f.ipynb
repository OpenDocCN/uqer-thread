```python
import talib
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
import quartz_futures as qf
from quartz_futures.api import *
```

布林带是美国股市分析家约翰·布林设计出来的一种非常简单实用的技术分析指标。一般而言,股价总是  
围绕某一价值来回上下波动，这一价值一般使用股票的历史平均价来表示。而布林带则利用统计学中的  
标准差原理，通过股票价格序列的标准差来衡量波动的幅度，而且股价通道的宽窄随着股价波动幅度的  
大小而变化，因此布林带自身也在不断地调整。  
布林带是由三条线构成，分别称之为上轨，中轨，下轨，具体的计算如下：  
**中轨** = MA(close, N)  
**上轨** = 中轨 + p\*std(close, N)  
**下轨** = 中轨 - p\*std(close, N)  
从公式来看，有点类似于我们统计学中置信区间的计算，这里的N我们取10天，而p相应的取1。

# 2.策略的逻辑
-------------
根据置信区间的理论，当股票的价格向上突破了上轨时，它有极大的可能是会重新回来的，所以此时  
做空便能盈利，向下突破也是相同的原理。具体的逻辑为：  
**开平仓:**  
- ### 若上一时刻收盘价上穿布林带的上轨，做空，若有多头则先平仓  
- ### 若上一时刻收盘价下穿布林带的下轨，做多，若有空头则先平仓

**止盈止损:**  
- ### 若当前收益超过0.01，止盈平仓  
- ### 若当前亏损超过0.01，止损平仓

我们来看看策略在这三个合约上具体的表现。

```python
all_perf = []
all_bt = []
backtest_universe = ['IH1612', 'IF1612', 'IC1612', 'rb1612', 'zn1612', 'cu1612']
multi= [300, 300, 200, 10, 5, 5]
for i in range(len(backtest_universe)):
    universe = [backtest_universe[i]]
    start = "2016-05-01"                 # 回测开始时间
    end   = "2016-08-05"                 # 回测结束时间
    capital_base = 1500000               # 初始可用资金
    refresh_rate = 1                     # 算法调用周期
    freq = 'd'
    commission = 0.3/10000
    len_window = 10

    def initialize(futures_account):
        futures_account.portfolio_value = []

    def handle_data(futures_account):
        symbol = get_symbol(universe[0])
        #print '\n'
        futures_account.portfolio_value.append(futures_account.cash+futures_account.position.get(symbol, dict()).get('long_margin', 0)+
                                           futures_account.position.get(symbol, dict()).get('short_margin', 0))
        #print futures_account.current_date, futures_account.current_time, symbol
        #print(futures_account.portfolio_value[-1])
    
        data = get_symbol_history(symbol=symbol, field='closePrice', time_range=len_window+1)
        closeprice_20 = np.array(data[symbol]['closePrice'], dtype='float')
        ma_20_1= talib.MA(closeprice_20,len_window)[-1]
        ma_20_2 = talib.MA(closeprice_20,len_window)[-2]
        std_20_1 = closeprice_20[1:].std()
        std_20_2 = closeprice_20[:-1].std()
    
        bolling_up_1 = ma_20_1+1*std_20_1
        bolling_down_1 = ma_20_1-1*std_20_1
        bolling_up_2 = ma_20_2+1*std_20_2
        bolling_down_2 = ma_20_2-1*std_20_2
        close_1 = closeprice_20[-1]
        close_2 = closeprice_20[-2]

        #print '布林线上通道:', bolling_up_1, ';', '布林线下通道：', bolling_down_1, ';', '昨日收盘：', close_1
        long_position = futures_account.position.get(symbol, dict()).get('long_position', 0)
        short_position = futures_account.position.get(symbol, dict()).get('short_position', 0)
        amount = int(futures_account.cash/(closeprice_20[-1]*multi[i]*0.3))
        if close_1 > bolling_up_1 and close_2 < bolling_up_1 and bolling_up_1 > bolling_up_2 and bolling_down_1 > bolling_down_2:
            if short_position < amount :
                #print futures_account.current_date, futures_account.current_time, '卖出开仓'
                order(symbol, -amount, 'open')
            if long_position > 0:
                #print futures_account.current_date, futures_account.current_time, '卖出平仓'
                order(symbol, -long_position, 'close')
        if close_1 <= bolling_down_1 and close_2 > bolling_down_2 and bolling_up_1 < bolling_up_2 and bolling_down_1 < bolling_down_2:
            if short_position > 0:
                #print futures_account.current_date, futures_account.current_time,'买入平仓'
                order(symbol, short_position, 'close')
            if long_position < amount:
                #print futures_account.current_date, futures_account.current_time, '买入开仓'
                order(symbol, amount, 'open')
        #print '仓位：', futures_account.position
        #print '现金：', futures_account.cash
        #print '组合价值：', futures_account.portfolio_value[-1]
    
        # 止盈止损
        if len(futures_account.portfolio_value) >= 2:
            returns = futures_account.portfolio_value[-1]/futures_account.portfolio_value[-2]
            #print(returns)
            if returns > 1.01:
                if long_position > 0:
                    #print futures_account.current_date, futures_account.current_time,'止盈：卖出平仓'
                    order(symbol, -long_position, 'close')
                if short_position > 0:
                    #print futures_account.current_date, futures_account.current_time,'止盈：买入平仓'
                    order(symbol, short_position, 'close')
            if returns < 0.99:
                if long_position > 0:
                    #print futures_account.current_date, futures_account.current_time,'止损：卖出平仓'
                    order(symbol, -long_position, 'close')
                if short_position > 0:
                    #print futures_account.current_date, futures_account.current_time,'止损：买入平仓'
                    order(symbol, short_position, 'close')
        
    bt, perf = qf.backtest.backtest(universe=universe, start=start, end=end,
                                initialize=initialize, handle_data=handle_data,
                                capital_base=capital_base, refresh_rate=refresh_rate,
                                freq=freq)
    all_perf.append(perf)
    all_bt.append(bt)
```
