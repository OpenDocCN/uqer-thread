# 试手1-商品期货跨品种择时套利策略-渤海证券研报

一. 跨品种套利介绍
    跨品种套利背后的逻辑出发点，既可以通过基本面分析得出，如两品种处于同一产业链的上下游或两品种具有一定的消费替代性，也可以通过统计分析得出，如两品种的价格走势具有较高的相关性同时满足协整关系。
    对商品期货跨品种套利的市场理解包括两个层面：一是如果构建一个跨品种组合，涉及到品种的选择和配比系数的确定；二是对价差波动方式的理解。

二. 品种选择和配比系数
    选择了 6 对组合，分别是：豆油-棕榈油、豆油-菜油、大豆-豆粕、螺纹钢-铁矿石、焦炭-焦煤、玉米-淀粉，其中玉米-淀粉和螺纹钢-铁矿石两对组合并不满足协整关系，但是具有明显的上下游关系。

```python
from __future__ import division
import quartz_futures as qf
from quartz_futures.api import *
from CAL.PyCAL import *
import numpy as np
import math
import datetime
import talib
import matplotlib.pyplot as plt 
import pandas as pd
from statsmodels.tsa.stattools import adfuller
import re

# 豆油-棕榈油期货价格走势图
data = DataAPI.MktMFutdGet(mainCon=1,contractObject=u"",startDate=u"20080102",endDate=u"20161208",field=u"",pandas="1")
data_p = data[data.contractObject=='P'].closePrice.reset_index(drop=True)
data_y = data[data.contractObject=='Y'].closePrice.reset_index(drop=True)
datep = data[data.contractObject=='P']['tradeDate']
datep = [datetime.datetime.strptime(x, "%Y-%m-%d").date() for x in datep]

fig, ax = plt.subplots(figsize=(12,6))
ax.plot_date(datep,data_p,'-',label="P Price")
ax.plot_date(datep,data_y,'--',label="Y Price")
ax.yaxis.grid()
plt.setp(plt.gca().get_xticklabels(),rotation=30)
plt.show()
```

```python
# 豆油期货主力合约日收盘价和棕榈油期货主力合约日收盘价分别进度单位根检验
dftest = adfuller(data_p)
dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
print 'P价格序列adf检验\n',dfoutput

dftest = adfuller(data_y)
dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
print '\nY价格序列adf检验\n',dfoutput
#  P值分为 0.1609 和 0.3090

# 对两个价格序列做一阶差分后再做单位根检验
dftest = adfuller(data_p.diff(1)[1:])
dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
print '\nP价格序列的一阶差分adf检验\n',dfoutput

dftest = adfuller(data_y.diff(1)[1:])
dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
print '\nY价格序列的一阶差分adf检验\n',dfoutput
# P 值为 3.46e-27 和 1.58e-21，均满足平稳性条件，表明两者是同阶的价格序列。
```

```python
# 回归分析
from sklearn import datasets,linear_model
from scipy import stats
slope, intercept, r_value, p_value, slope_std_error = stats.linregress(data_p.values, data_y.values)
predict_y = intercept + slope * data_p.values
pred_error = data_y.values - predict_y
degrees_of_freedom = len(data_p.values) - 2
residual_std_error = np.sqrt(np.sum(pred_error**2) / degrees_of_freedom)
print '\nslope：',slope,'\n','intercept:',intercept,'\n'
dftest = adfuller(pred_error)
# 对上述函数求得的值进行语义描述
dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
print 'dfoutput\n',dfoutput
# 对 PY =a+ b*PP做回归分析可得 PY = 939.4+1.017*PP，同时对其残差进行单位根检验，其P值为 0.0189，不具有单位根，表现 PY和 PM具有协整关系。可得到价差序列spread= PY- PP。
```

三. 价差波动分析
对具有较明显的波动边界的走势，使用灵敏的 KD 技术指标做开仓信号的处理；对没有明显波动边界的走势，使用自适应均线来做开仓信号的处理。
1.KD 指标介绍
    假设spread 为价差序列， rsv 指标为当前价差与过去 N 天最小值的差与过去 N 天最大值与最小值差的比例，在上涨行情中该值趋近于 1，在下跌行情中该值趋近于 0，K 指标为 rsv 过去 m 天的均值， D 指标为 K 指标过去 m 天的均值。
    rsv_t = [ spread_t - min(spread_t-N+1 , spread_t) ] / [ max(spread_t-N+1 , spread_t) - min(spread_t-N+1 , spread_t) ]
    K_t = mean(rsv_t-m+1,rsv_t)
    D_t = mean(K_t-m+1,K_t)
2.自适应均线介绍
    dl表示 n 天的净价格变动， vt表示 n 天价格变化的综合，揭示市场噪音总量。 n 根据不同品种组合的价差均值回复时的特征来选择.
    d_l = |p_l – p_l-n|，
    v_t = ∑_|t – t-n| |p_l - p_l-1|，
    ERt=dt/vt
    smooth = ER*(fastest-slowest)+slowest,
    fastest = 2/(Nf+1),slowest = 2/(Ns+1),
    c=smooth*smooth
    AMA_t = AMA_t-1 + c*(spread_t - AMA_t-1)
    
四.交易信号介绍
1.开仓信号
	如果历史上极值点的分布比较稳定，则使用 KD 指标来做进行信号的判断。
	当价差大于一定水平，同时 KD 指标的 K 线向下穿越 D 线时，开仓做空价差；
	当价差小于一定水平，同时 KD 指标的 K 线向上穿越 D 线时，开仓做多价差。

	如果历史上极值点的分布不是很稳定，而是在一个较大的范围内波动，则使用自适应均线来做信号的判断。
	当价差大于正常波动范围的上轨，同时自适应均线向下拐头时，即 AMAt-1 < AMAt-2且 AMAt-2 > AMAt-3时，做空价差；
	当价差小于正常波动范围的下轨，同时自适应均线向上拐头时，即 AMAt-1 > AMAt-2且 AMAt-2 < AMAt-3时，进场做多价差。
    
2.平部分仓位信号
	做空价差后，在价差未回归到正常波动水平时，如果 AMAt-1 > AMAt-2 且AMAt-2 < AMAt-3时，先平掉部分仓位；
	做多价差后，在价差未回归到正常波动水平时，如果 AMAt-1 < AMAt-2且 AMAt-2 > AMAt-3时，先平掉部分仓位。
    
3.止损信号
	对于使用 KD 指标进场的套利仓位，
	如果是做空价差的仓位，那么在价差创出历史新高后止损平仓；
	如果是做多价差的仓位，那么在价差创出历史新低后止损平仓。
    
	对于使用自适应均线做为信号开仓的仓位，
	如果是做空价差的仓位，那么当自适应均线的值大于开仓时自适应均线的值时，止损平仓；
	如果是做多价差的仓位，那么当自适应均线的值小于开仓时自适应均线的值时，止损平仓。
4.止盈信号
	进场做空价差后，如果价差回落小于 Ma+CP_Short *Std 时，则 落 入 止 盈 区 间 止 盈 平 仓 ； 
	进场做多价差后，如果价差上涨大于 Ma+CP_Long *Std时，则落入止盈区间止盈平仓
    
五. 单个组合回测
	豆油-棕榈油组合的做多价差套利使用较敏感的 KD 指标择时法进行交易，做空价差套利使用自适应均线择时的方法来进行交易。	

```python
from __future__ import division
import quartz_futures as qf
from quartz_futures.api import *
from CAL.PyCAL import *
import numpy as np
import math
import datetime
import talib
import matplotlib.pyplot as plt 
import pandas as pd
from statsmodels.tsa.stattools import adfuller
import re
import numpy as np

def KDindex(spread,N,m):  #KD指标计算
    kt = []
    for s in range(1,m+1):
        rsv = []
        for t in [i+s-1 for i in range(1,m+1)]:   
            spt = spread[-t-N+1:-t+1] if t!=1 else spread
            rsvt = (spt[-1] - min(spt[-N:]))/(max(spt[-N:]) - min(spt[-N:]))
            rsv.append(rsvt)
        kt.append(np.mean(rsv))
    dt=np.mean(kt)
    return(kt[0],dt)


def AMA(spread,n,Ns=60,Nf=4): #AMA指标计算
    c = []
    fastest = 2/(Nf+1)
    slowest = 2/(Ns+1)
    for i in range(len(spread)):
        try:
            dl = abs(spread[-1-i]-spread[-n-1-i])
            if i==0:
                vt = sum([abs(x) for x in np.diff(spread[-n-1-i:])])
            else:
                vt = sum([abs(x) for x in np.diff(spread[-n-1-i:-i])])
            ERt = dl/vt
            smooth = ERt * (fastest - slowest) +slowest
            ci = smooth**2
        except:
            ci = 0
        c.insert(0,ci)
    AMA = len(c)*[0]
    AMA[0] = spread[0]
    for i in range(1,len(AMA)):
        AMA[i] = AMA[i-1]+c[i]*(spread[i]-AMA[i-1])
    return AMA
```
