# 波动率研究之一_GARCH模型

一、引言
---
由于资产配置、风险管理、资产定价和量化投资策略等都与波动率有关，资产、策略与因子的波动率的测量和预测是量化投资的最为重要的议题之一。多年来，国内外学术界和实践界对波动率测量和预测方法进行了大量研究，波动率测量和预测方法众多，本期开始尝试对波动率测量和预测方法进行学习研究。
  
二、波动率的定义
--
  波动率是指变量随时间序列而呈现的扰动，是对资产、策略和因子收益不确定性和风险的衡量指标。为了准确度量波动率，由简到繁，大量统计、计量模型被逐步被用于构建波动率模型。
  
三、波动率的类型
--
  1、历史波动率
  2、隐含波动率
  3、现实波动率（高频数据波动率）
  
四、波动率的特征
--
  1、尖峰厚尾特征
  2、波动集群性
  3、杠杆效应特征
  4、信息流特征
  5、长期记忆与持续性
  6、共生波动特征
  7、溢出效应特征
  8、微笑现象
  9、隐含的相关波动
  
五、波动率的预测
--
   1、 模型、研究众多，不断改进创新；
   2、仍是金融工程和计量经济学的前沿学科的研究对象。
     具体参考文献：
     1、陈浪南等，《波动率研究》，中国财经出版社，2008年08月；
     2、李 隽，关于波动率及其预测方法的文献综述，山西青年管理干部学院学报，2009年33月；



六、GARCH模型
-----
      传统的计量经济学对时间序列变量的第二个假设：假定时间序列变量的波动幅度（方差）是固定的，不符合实际，比如，人们早就发现股票收益的波动幅度是随时间而变化的，并非常数。这使得传统的时间序列分析对实际问题并不有效。罗伯特·恩格尔在1982年发表在《计量经济学》杂志（Econometrica）的一篇论文中提出了ARCH模型解决了时间序列的波动性（volatility）问题。
     GARCH模型称为广义ARCH模型，是ARCH模型的拓展，由Bollerslev(1986)发展起来的。方差用ARMA模型来表示。GARCH模型提出来以后，在它基础上又出现了非常多的扩展模型。这些扩展模型大多数是对GARCH有关条件的改变。
     本文就主要就恒生指数的波动率用GARCH模型分析下，同时恒指波幅对比分析

```python
#取出深300指数、创业板指、国债指数、企业债指数、恒生指数、标普500指数、原油指数、黄金指数数据，并进行合并
start = '20130501'
end = '20180531'
indices = ['000300', '399006', '000012','000013','HSI','SPX']
df = DataAPI.MktIdxdGet(ticker=indices, beginDate=start, endDate=end, field="tradeDate,ticker,closeIndex")
df = df.pivot(index="tradeDate", columns="ticker", values="closeIndex")
df.head()
```

```python
#计算各指数收益率数据
df1=df.pct_change()
df_returns=df1.dropna()
df_returns.head()
```

```python
#计算沪深300指数GARCH模型
returns = 100.0 *df_returns['000300']
am2 = arch_model(returns)
res2 = am2.fit()
res2.summary()
```

```python
res2.resid.plot()
```

```python
res2.conditional_volatility.plot()
```

```python
returns = 100.0*df_returns['HSI']
am3 = arch_model(returns, p=1, q=1)
res3 = am3.fit()
fig = plt.figure()
fig = returns.plot()
res3.summary()
```

```python
res3.resid.plot()
```
