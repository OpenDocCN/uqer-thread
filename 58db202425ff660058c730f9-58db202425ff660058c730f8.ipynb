```python
import numpy as np
import pandas as pd
from math import *
```

```python
secID1=DataAPI.SecIDGet(partyID=u"",assetClass=u"",ticker=u"000379,000759",cnSpell=u"",field=u"secID",pandas="1")
secID2=DataAPI.SecIDGet(partyID=u"",assetClass=u"",ticker=u"000549,000577,040036,700005,",cnSpell=u"",field=u"secID",pandas="1")
secID1=secID1.drop(secID1.index[2])
secID1=secID1.ix[:, 0].tolist()
secID2=secID2.drop(secID2.index[1])
secID2=secID2.ix[:, 0].tolist()
```

```python
def get_covmat(secID1,secID2,date):
    '''
    输入secid + 日期，获得以此估计出来的年化方差协方差矩阵
    '''
    start_date = str(int(date[:4])-3) + date[4:]
    Monetary_Fund=DataAPI.FundDivmGet(dataDate=u"",secID=secID1,ticker=u"000379,000759",beginDate=start_date,endDate=date,field=u"",pandas="1")
    Monetary_Fund=Monetary_Fund.pivot(columns='secID',index='endDate',values='weeklyYield')
    Monetary_Fund=pd.DataFrame(Monetary_Fund)
    Monetary_Fund['000379.OFCN']=Monetary_Fund['000379.OFCN'].apply(lambda X:pow(X,1./365)-1)
    Monetary_Fund['000759.OFCN']=Monetary_Fund['000759.OFCN'].apply(lambda X:pow(X,1./365)-1)
    Stock_Bond_Fund=DataAPI.FundNavGet(secID=secID2,ticker=u"",beginDate=start_date,endDate=date,field=u"secID,endDate,ADJUST_NAV",pandas="1")
    Stock_Bond_Fund=Stock_Bond_Fund.pivot(index='endDate',columns='secID', values='ADJUST_NAV').pct_change().fillna(0.0)
    Stock_Bond_Fund=pd.DataFrame(Stock_Bond_Fund)
    return_mat=pd.merge(Stock_Bond_Fund,Monetary_Fund, how = 'left', left_index = True, right_index=True).fillna(0.0)
    return return_mat.cov()*250
```

```python
date = '20160101'
cov_mat = get_covmat(secID1,secID2,date)
cov_mat
```

```python
def get_weights(secID1,secID2, date):
    '''
    secid + 日期，获得以此估计出来的年化方差协方差矩阵
    '''
    cov_mat = get_covmat(secID1,secID2, date)
    # 根据上面的描述写好对应的P、q、G、h、A、b
    Nums = cov_mat.shape[0]
    P = matrix(cov_mat.values)
    q = matrix(np.zeros(Nums))
    G = matrix(np.diag(-np.ones(Nums)))
    h = matrix(np.zeros(Nums))
    A = matrix(np.ones(Nums)).T
    b = matrix(1.0).T
    
    # 代入优化方程求解
    solvers.options['show_progress'] = False
    sol = solvers.qp(P, q, G, h, A, b)
    wts = pd.Series(index=cov_mat.index, data=np.array(sol['x']).flatten())
    wts = wts[wts >= 0.0001]  # weights调整，对于权重太小的不予配置
    wts_adjusted = wts / wts.sum() * 1.0
    return wts_adjusted
```
