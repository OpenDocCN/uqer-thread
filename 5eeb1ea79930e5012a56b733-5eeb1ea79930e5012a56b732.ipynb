# 总理-净利润同比增长

```python
import numpy as np
import pandas as pd

universe=set_universe("A")
year_2017='2017'
lastYear='2018'
thisYear="2019"

# 净利润表-2017（年度）
data0=DataAPI.FdmtISGet(ticker='',secID=universe,reportType=u"A",endDate=u"",beginDate=u"",publishDateEnd=u"",publishDateBegin=u"",endDateRep="",beginDateRep="",
                  beginYear=year_2017,endYear=year_2017,fiscalPeriod="",field=u"ticker,NIncomeAttrP,secShortName,endDateRep",pandas="1")

# 净利润表-2018（年度）
data1=DataAPI.FdmtISGet(ticker='',secID=universe,reportType=u"A",endDate=u"",beginDate=u"",publishDateEnd=u"",publishDateBegin=u"",endDateRep="",beginDateRep="",
                  beginYear=lastYear,endYear=lastYear,fiscalPeriod="",field=u"ticker,NIncomeAttrP,secShortName,endDateRep",pandas="1")

# 净利润表-2019（年度）
data2=DataAPI.FdmtISGet(ticker='',secID=universe,reportType=u"A",endDate=u"",beginDate=u"",publishDateEnd=u"",publishDateBegin=u"",endDateRep="",beginDateRep="",
                  beginYear=thisYear,endYear=thisYear,fiscalPeriod="",field=u"ticker,NIncomeAttrP,secShortName,endDateRep",pandas="1")

data0=data0.groupby(["ticker"],as_index=False).first()
data1=data1.groupby(["ticker"],as_index=False).first()
data2=data2.groupby(["ticker"],as_index=False).first()

data=pd.merge(data0[['ticker','secShortName','NIncomeAttrP']],data1[['ticker','NIncomeAttrP']], on='ticker',how='inner')
data=pd.merge(data,data2[['ticker','NIncomeAttrP']], on='ticker',how='inner')
data['growth_1%']=(data['NIncomeAttrP_y']-data['NIncomeAttrP_x'])/data['NIncomeAttrP_x']*100
data['growth_2%']=(data['NIncomeAttrP']-data['NIncomeAttrP_y'])/data['NIncomeAttrP']*100

data=data[(data['growth_1%']>10)& (data['growth_1%']<20)&(data['growth_2%']>10)& (data['growth_2%']<20)]
data.sort_values(['growth_2%','growth_1%'],ascending=[False,False], inplace=True)
data.rename(columns={'NIncomeAttrP_x':u'第1年利润',
                                       'NIncomeAttrP_y':u'第2年利润', 
                                        'NIncomeAttrP':u'第3年利润'},inplace=True)
data
```

```python
import numpy as np
import pandas as pd

# 商誉减值公告 000100 TLiab=总负债 TAssets=资产总计 净资产=总资产-总负债 归母净资产=净资产-少数所有者权益
universe=set_universe("A")
thisYear="2019"

# 合并资产负债表(3季度)
data = DataAPI.FdmtBSGet(ticker=u"",secID=universe,reportType=u"Q3",endDate=u"",beginDate=u"",
                  publishDateEnd=u"",publishDateBegin=u"",endDateRep="",beginDateRep="",beginYear=thisYear,endYear=thisYear,fiscalPeriod="",
                         field=u"ticker,goodwill,TAssets,TLiab,minorityInt",pandas="1")
data=data.dropna()
data=data.drop_duplicates()

# row=data.iloc[-1]
# # 商誉
# print(u"商誉="+str(row.goodwill))
# # 净资产（归母净资产）
# print(u"归母净资产="+str(row.TAssets-row.TLiab-row.minorityInt))
# # 商誉占归母净资产的比重 当比例大于30% (0.3) 时注意商誉风险
# print(u"商誉占净资产比重="+str(row.goodwill/(row.TAssets-row.TLiab-row.minorityInt)))

# 净利润表（3季度）
data1=DataAPI.FdmtISGet(ticker=u"",secID=universe,reportType=u"CQ3",endDate=u"",beginDate=u"",publishDateEnd=u"",publishDateBegin=u"",endDateRep="",beginDateRep="",
                  beginYear=thisYear,endYear=thisYear,fiscalPeriod="",field=u"ticker,NIncomeAttrP,secShortName",pandas="1")

# data1=data1.dropna()
# data1=data1.drop_duplicates()
# row1=data1.loc[-1]
# # 归母净利润
# print(u"归母净利润="+str(row1.NIncomeAttrP))

data2=data.merge(data1, on="ticker")

data2[u"商誉占净资产比例"]=np.round(data2["goodwill"]/(data2.TAssets-data2.TLiab-data2.minorityInt),2)
data2[u"商誉占净利润比例"]=np.round(data2["goodwill"]/data2.NIncomeAttrP,2)

# 排序
data3=data2[(data2[u"商誉占净利润比例"]>=1)|(data2[u"商誉占净资产比例"]>=0.3)].sort_values(by=[u"商誉占净资产比例",u"商誉占净利润比例"],ascending=[False,False])
print(u"扫到的雷-前10")
data3.head(50)
```
