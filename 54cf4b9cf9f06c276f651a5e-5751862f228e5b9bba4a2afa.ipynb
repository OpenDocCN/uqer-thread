# 波动率曲面

在本文中，我们将通过实际的市场的例子，展示如何在量化实验室中计算和展示期权的隐含波动率微笑。

1. 数据准备
-----------------

在本节中，我们首先读入原始数据文件 [期权行情_20150127.xlsx](http://pan.baidu.com/s/1pJumtkv) ，进行一些必要的数据转换。

* `tradeDate` 交易日
* `spot` 标的上证50ETF的收盘净值
* `optData['TTM']` 期权到期时间（单位为年）
* `optData['STRIKE']` 期权行权价格
* `optData['SETTL_PRICE']` 期权当日结算价格
* `optDat['CONTRACT_TYPE']` 期权类型，看涨或者看跌

```python
import datetime as dt
optData = pd.read_excel(u'期权行情_20150127.xlsx', 'Sheet1')
```

```python
tradeDate = dt.date(2015,1,27)
riskFree = 0.05

spot = 2.483
optData['TTM'] = ((optData['EXP_DATE']- tradeDate).apply(lambda x:float(x))  /24/3600/1e9 + 1)/ 365.0
optData['Type'] = Option.Call
optData['Type'][optData['CONTRACT_TYPE'] == 'PO'] = Option.Put
```

2. 计算隐含波动率
--------------------

接着我们可以方便的使用内置函数 `BSMImpliedVolatity` 计算期权的隐含波动率。

```python
impliedVol = BSMImpliedVolatity(optData['Type'], optData['STRIKE'], spot, riskFree, 0.0, optData['TTM'], optData['SETTL_PRICE'])
impliedVol['strike'] = optData['STRIKE']
impliedVol['expiry'] = optData['EXP_DATE']
impliedVol['type'] = optData['CONTRACT_TYPE']
```

```python
res = impliedVol.groupby(['type', 'expiry'])
```

我们分别计算看涨以及看跌期权的隐含波动率，并计算它们的平均，保存在 `dataSet['Avg']` 中。

```python
import numpy as np
optionTypes= ['CO', 'PO']
expiries = impliedVol['expiry'].unique()
maturity = optData['TTM'].unique()
strikes = impliedVol['strike'].unique()

dataSet = {}

for t in optionTypes:
    resArray = np.ndarray((len(strikes), len(expiries)))
    for i, e in enumerate(expiries):
        sample = res.get_group((t,e))
        resArray[:,i] = sample['vol']
    dataSet[t] = resArray
dataSet['Avg'] = (dataSet['CO'] + dataSet['PO']) / 2.0
```

配合 `pandas` 的绘图功能，可以直观的观察波动率微笑在不同的期限下的表现。

```python
volTermStructure = pd.DataFrame(dataSet['Avg'], index = strikes, columns=expiries)
ax= volTermStructure.plot(figsize=(16,8))
ax.set_xlabel('Strike', fontdict={"size":15})
ax.set_ylabel('Volatility', fontdict={"size":15})
```

3. 构造波动率曲面
----------------------

但是对于市场参与者而言，像刚才这样仅仅观察的线的结构不够。他们需要看到整个市场以到期时间，行权价为轴的波动率曲面（Volatility Surface）。除此之外，他们更想知道，波动率曲面上，那些并不是市场报价点的值，至少是个估计。这样的波动率曲面构造，往往需要依赖某种模型，或者某种插值方法。在这一节中，我们将介绍使用 `CAL` 中的波动率曲面构造函数。

3.1 一些数据准备
------------------

我们首先计算一些后面需要使用的参数，例如：

* `forwards` 标的物的远期价格
* `strikeGrid` 行权价格方向网格
* `maturityGrid` 到期时间网格

```python
from math import exp
maturityDate = [Date(2015,1,28), Date(2015,2,25), Date(2015,3,25), Date(2015,6,24)]
forwards = [spot * exp(x*riskFree) for x in maturity]

strikeGrid = np.linspace(strikes[0], strikes[-1], 200)
maturityGrid = np.linspace(maturity[0], maturity[-1], 200)
```

3.2 基于SABR模型的波动率曲面
----------------------------

这个例子基于 `CAL` 函数 `SABRCalibratedVolSruface`

```python
volInterpolator = SABRCalibratedVolSruface(strikes, forwards,maturity, dataSet['Avg'].T, interp= 'Bilinear')
volSurface = np.ndarray((len(strikeGrid), len(maturityGrid)))
for i, s in enumerate(strikeGrid):
    for j, m in enumerate(maturityGrid):
        volSurface[i,j] = volInterpolator.volatility(m,s)
```

```python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm

maturityMesher, strikeMesher = np.meshgrid(maturityGrid, strikeGrid)
fig= pylab.figure(figsize = (16,8))
ax = fig.add_subplot(1, 1, 1, projection = '3d')
surface = ax.plot_surface(strikeMesher, maturityMesher, volSurface*100, cmap=cm.jet)
fig.colorbar(surface,shrink=0.75)
ax.set_xlabel("strike", fontdict={"size":15})
ax.set_ylabel(r"maturity", fontdict={"size":15})
ax.set_title('SABR Volatility Surface',fontdict={"size":20})
```

3.2 基于SVI模型的波动率曲面
------------------------------

这个例子基于 `CAL` 函数 `SviCalibratedVolSruface`

```python
volInterpolator = SviCalibratedVolSruface(strikes, forwards,maturity, dataSet['Avg'].T, interp= 'Bilinear')
volSurface = np.ndarray((len(strikeGrid), len(maturityGrid)))
for i, s in enumerate(strikeGrid):
    for j, m in enumerate(maturityGrid):
        volSurface[i,j] = volInterpolator.volatility(m,s)
        
maturityMesher, strikeMesher = np.meshgrid(maturityGrid, strikeGrid)
fig= pylab.figure(figsize = (16,8))
ax = fig.add_subplot(1, 1, 1, projection = '3d')
surface = ax.plot_surface(strikeMesher, maturityMesher, volSurface*100, cmap=cm.jet)
fig.colorbar(surface,shrink=0.75)
ax.set_xlabel("strike", fontdict={"size":15})
ax.set_ylabel(r"maturity", fontdict={"size":15})
ax.set_title('SVI Volatility Surface',fontdict={"size":20})
```
