大家在回测的过程中用的股票池一般是当前市场的全部A股

```python
universe = set_universe('A')
print '300372.XSHE' in universe
print '600656.XSHG' in universe

universe = StockScreener(Factor.LCAP.nsmall(2000))
universe = universe.preview('2016-05-20')
print '300372.XSHE' in universe
```

可以看到今年退市的欣泰电气300372.XSHE和博元投资600656.XSHG都不在股票池中，大家都知道股票在退市之前经过大跌市值肯定是很低的，如果我们在做小市值回测时直接把它们从股票池中去掉显然是未来函数，有同学说退市之前会被ST，我们的股票池在回测去掉ST时会去掉这些退市股，首先我们看看踢掉ST能不能踢掉退市股

```python
st = DataAPI.SecSTGet(beginDate=u"20160101",endDate=u"20161021",secID=universe,ticker=u"",field=u"secID,tradeDate",pandas="1")
print st[st['secID'] == '300372.XSHE']
print st[st['secID'] == '600656.XSHG']
```

可以看到，SecSTGet的返回结果并没有博元投资和欣泰电气，也就是这些退市股不在ST列表中

而且以欣泰电气为例，在2016-05-20时欣泰电气并没有被ST，之后停牌，2016-07-21复牌时才被ST，也就是实盘的过程中我们可能会在2016-05-20选中欣泰电气，而回测的过程中我们已经莫名其妙的把它踢掉了

那如果我们在股票池中加入这些退市股是不是就可以避免这种情况，以600656.XSHG博元投资为例，每日买进1000股
