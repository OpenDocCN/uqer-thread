# 中国股市回报偏度研究
随着金融市场的发展，不同基于有效市场的理论被提出，如资本资产定价CAPM模型、期权定价B-S模型。但是很多结果表明，股票回报不是正态分布，而是有一定的偏度。本帖将探讨大盘指数回报的偏度、行业和个股回报的偏度情况。


```python
#!/usr/bin/python
#-*-coding:utf-8-*-

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random
```

```python
begin_date = u"20000101"
end_date = u"20180101"

def get_skewness(begin_date, end_date, stock_list):
    # 计算每天收益率
    daily_return = pd.DataFrame()
    for code in stock_list:
        data = DataAPI.MktIdxdGet(indexID=u"", ticker=code, tradeDate=u"", beginDate=begin_date, 
                                  endDate=end_date, exchangeCD=u"XSHE,XSHG", field=u"",pandas="1")
        data.set_index('tradeDate', inplace = True)
        data['return_temp'] = data['closeIndex'].pct_change() 
        
        # 去极值处理
        data['return'] = np.where(data['return_temp']>0.099, 0.099, data['return_temp'])
        data['return'] = np.where(data['return_temp']<-0.099, -0.099, data['return']) 
        daily_return[code] = data['return']
        
    i = 0
    for code in daily_return.columns:
        # 计算收益率的平均值、方差、偏度和峰度
        ret=daily_return[code].dropna()
        mean = ret.mean()
        std = ret.std()
        skewness = ret.skew()
        kurtosis = ret.kurt()
        # 画收益率的直方图
        info="%s :mean=%.5f  std=%.5f  skewness=%.5f  kurtosis=%.5f"%(code,mean,std,skewness,kurtosis)
        plt.subplot(len(daily_return.columns)*100+11+i)
        plt.xlabel("return")
        plt.ylabel("frequency")
        plt.title(info)
        plt.hist(np.array(ret), 200, normed=True) 
        plt.grid(True)
        plt.show()
        i += 1    
```

## 1 指数回报偏度研究。
------
	研究标的：上证指数，深圳成指，沪深300指数。
	时间区间：20000101-20180101。
	结果显示：三个指数收益率的偏度为负，呈现左偏现象，负收益率比正收益率的数据要多。峰度大于3，表明指数收益率分布比正态分布要陡峭。

```python
# 上证指数000001，深圳成指399001，沪深300指数000300
stock_list = [u"000001", u"399001", u"000300"]
get_skewness(begin_date, end_date, stock_list)
```

## 4 总结
------
         从上面的分析得出中国股市指数和行业回报偏度值一般为负值，低于均值的回报率偏多，幅度偏大，风险也大。原因如下：
      1、偏度现象可能由公司传递的正面和负面消息分布情况所产生的。在中国，公司的正负面消息存在不对称性，大多公司通常倾向于更多地向市场传递正面消息而不是负面消息，但是负面消息一经传递便会成簇的形式释放，从而 使得资产收益率分布呈现出左偏现象。
      2、依照期望理论，投资者的看法主观地依赖他们的参考点，大多数人在面临损失时是风险爱好，而面对收益时风险厌恶。但当信息不确定时，投资者面对损失不再那么风险爱好，面对收益也不再风险厌恶。尤其是经济形势不好 时，人们对好消息开始怀疑，对坏消息期望更低。也就是或市场信息的不确定性使得市场低迷时投资者期望趋向放大负面信息而忽视正面信息，表现出左偏现象。
