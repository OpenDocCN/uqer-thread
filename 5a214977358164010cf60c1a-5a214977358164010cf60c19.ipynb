```python
def plot_under_water(bt):
    bt1 = bt
    bt_quantile_ten = bt1
    data = bt_quantile_ten[[u'tradeDate',u'portfolio_value',u'benchmark_return']]
    data['portfolio_return'] = data.portfolio_value/data.portfolio_value.shift(1) - 1.0
    data['portfolio_return'].ix[0] = data['portfolio_value'].ix[0]/	10000000.0 - 1.0
    data['excess_return'] = data.portfolio_return - data.benchmark_return
    data['excess'] = data.excess_return + 1.0
    data['excess'] = data.excess.cumprod()

    df_cum_rets = data['excess']
    running_max = np.maximum.accumulate(df_cum_rets)
    underwater = -((running_max - df_cum_rets) / running_max)
    underwater.index = data['tradeDate']

    fig = plt.figure(figsize=(12, 5))
    fig.set_tight_layout(True)
    ax1 = fig.add_subplot(111)
    ax2 = ax1.twinx()
    x = range(len(underwater))
    ax1.grid(True)
    ax1.set_ylim(-0.30, 0)
    ax1.fill_between(underwater.index, 0, np.array(underwater), color='#000066', alpha=1)

    ax2.plot(data['tradeDate'], data[['excess']], label='hedged(right)', color='r')
    ax2.set_ylim(bottom=0.9, top=3)
    s = ax1.set_title(u"股票月度调仓走势", fontproperties=font, fontsize=16)
    return fig

# 去极值，标准化，中性化
def pretreat_factor(factor_df, neu=True):
    pretreat_data = factor_df.copy(deep=True)
    for dt in pretreat_data.index:
        try:
            factor_dt = pretreat_data.ix[dt].dropna()
            factor_dt_dict = factor_dt.to_dict()
            if neu:
                pretreat_data.ix[dt] = pd.Series(neutralize(winsorize(factor_df.ix[dt].to_dict()), target_date=''.join(dt.split('-')), risk_module='short', industry_type='SW1',                                                             exclude_style_list=['BETA', 'RESVOL', 'MOMENTUM', 'EARNYILD', 'BTOP', 'SIZENL', 'GROWTH', 'LEVERAGE', 'LIQUIDTY']))
            else:
                pretreat_data.ix[dt] = pd.Series(standardize(winsorize(factor_dt_dict)))
        except Exception as excp:
            print dt
            print excp
            continue
    return pretreat_data
```

## 2.1 从行业中性说起
* 本文参考自**《FOF体系&多因子体系——东北证券金融工程2018年度策略报告》**，旨在提供一个因子构建的新思路。
* 我们在生成信号之前，通常先要对因子进行处理，包括去极值，中性化以及标准化。关于[**中性化**](https://uqer.datayes.com/community/share/59c108340f66ae010861b862)，我之前的帖子有提过。
* 传统的行业中性的做法，是因子对行业哑变量（一级行业分类）进行回归取残差，本质上是对同一个行业内的股票的因子进行中性化。
* 这么做的目的是在于，某些因子例如财务，估值，这些和行业市值密切相关的因子，不同行业之间是不能直接比较的，而行业中性可以使不同行业之间的因子值具有可比性，必要的时候，我们还应进行风格中性。
* 当然，我们不否认这样做的合理性，从历史来看，也没有什么问题。但肯定是有一些弊端，例如，这样的做法不能在更加细分的子行业进行区分。
* 优矿[**neutralize**](https://uqer.datayes.com/help/quartzExt/#neutralize)函数可以做上面这些事情

* 那么是不是深入到更细分领域做更细分的行业中性就能解决这个问题了呢？理论上可以。但是细分行业下，有些子行业的样本数量不足，导致在做平均值时受到极端情况影响较大，统计上意义也不大。
* 我们转换思路，如果横截面比较比较困难，那么从时间序列的角度来探究是否可行？
* 重点来了，如果我们认可‚“市场普遍对于大类行业及其细分子行业存在估值中枢”，那么这个估值中枢就能帮我们加以判断某个股票现在是处于估值中枢之上（贵）还是之下（便宜）。因此，只要这个公司的主营业务没有发生巨大变化，那么其估值水平相对自身历史就是可比的。借助这个可比性，我们可以剔除不同行业之间的影响：**传统的横截面行业中性分析得出相对行业平均水平的贵与便宜，而我们从时间序列分析得出相对于自身的贵与便宜。对于这种在时间序列维度上分析构建而成的因子我们称之为择时因子，或者也可以把运用择时思想构建的因子称之为择时因子，其本质是一种因子的因子（Factor of Factor）的做法。**
* 之前，少博曾在社区发布[**相对市盈率**](https://uqer.datayes.com/community/share/586dfe7089e3ba004eefe2e5)这个因子，有兴趣的读者可以比较下它们之间的区别（时间序列 vs 横截面）
* 还可以构造相对行业市盈率因子（感觉市盈率都要被玩坏了）

## 2.3 因子的计算
* 衡量时间序列上的高和低的方法很多，我们这里仅选择较为简单的一种方法：分位数法，即计算过去一段时间序列上的当前因子值分位数，把因子的绝对值映射成[0,1]区间的百分数。这里时间周期我们选择过去3年、5年、10年。不选择全历史的原因在于不同股票起始时间不同，无法保证时间段的可比性。
* 构建完毕择时因子之后，我们希望选出那些相对自身而言被高估的股票做空，相对自身而言被低估的股票做多。然后这样的思路是否能够带来超额收益。
* 在构建估值因子的分位数时，所用的是PE的倒数EP，PE均指的是TTM。
* 过去一段时间至少是3年。实际上在计算中是用到过去3\*240个非停牌交易日数据进行计算。
* 我又要来科普了，分位数法，在多因子模型里头，还有一个地方可以用到。当大家给因子做标准化的时候，想把因子的分布转换成标准正态分布，然后，很多因子没法转，这时候，一个办法是强制使用正态分布的逆函数来转，怎么做，这时候可以借助**分位数标准化**的方法？先排序，把因子值所在的分位数放在正态分布的概率分布的逆函数中去，这样得到的因子值，是服从正态分布的。有一个问题在于，这样做只使用了股票之间因子值的顺序的信息，没有考虑股票之间因子的距离，对于多因子模型，还是有些影响，完毕。
* 那么这个因子会把上市不足3\*240天的股票剔除掉，覆盖率需要去考虑

## 3.1 因子的预测能力

```python
# 计算了每一个月的**因子**和**之后20日收益**的秩相关系数
def get_ic(factor, price, title):
    ic_data = pd.DataFrame(index=factor.index, columns=['IC','pValue'])

    # 计算相关系数
    for dt in ic_data.index:
        tmp_factor = factor.ix[dt]
        tmp_ret = price.ix[dt]
        cor = pd.DataFrame(tmp_factor)
        ret = pd.DataFrame(tmp_ret)
        cor.columns = ['corr']
        ret.columns = ['ret']
        cor['ret'] = ret['ret']
        cor = cor[~pd.isnull(cor['corr'])][~pd.isnull(cor['ret'])]
        if len(cor) < 5:
            continue

        ic, p_value = st.spearmanr(cor['corr'],cor['ret'])   # 计算秩相关系数 RankIC
        ic_data['IC'][dt] = ic
        ic_data['pValue'][dt] = p_value

    print 'mean of IC: ', ic_data['IC'].mean(), '；',
    print 'median of IC: ', ic_data['IC'].median()
    print 'the number of IC(all, plus, minus): ', (len(ic_data), len(ic_data[ic_data.IC>0]), len(ic_data[ic_data.IC<0]))


    # 每一天的**因子**和**之前20日收益**的秩相关系数作图
    ic_data.index = pd.to_datetime(ic_data.index)
    fig = plt.figure(figsize=(16, 6))
    ax = fig.add_subplot(111)

    lns = ax.plot(ic_data[ic_data>0].index, ic_data.IC, label='IC')

    labs = [l.get_label() for l in lns]
    ax.legend(lns, labs, bbox_to_anchor=[0.6, 0.1], loc='', ncol=2, mode="", borderaxespad=0., fontsize=12)
    ax.set_ylabel(u'相关系数', fontproperties=font, fontsize=16)
    ax.set_xlabel(u'日期', fontproperties=font, fontsize=16)
    ax.set_title(title, fontproperties=font, fontsize=16)
    ax.grid()
    return ic_data
ic_data = get_ic(factor_pure, forward_20d_return_data, u'择时因子与之后一个月收益月度IC')
```

从上面计算结果可知，IC的均值大概为0.05， 该因子和之后20日收益的秩相关系数在大部分时间为正，因子对之后一个月的收益具有预测性。

## 3.2 分组超额收益

```python
n_quantile = 10
# 统计十分位数
cols_mean = [i+1 for i in range(n_quantile)]
cols = cols_mean

excess_returns_means = pd.DataFrame(index=factor_pure.index, columns=cols)

# 计算因子分组的超额收益平均值
for dt in excess_returns_means.index:
    qt_mean_results = []
    
    # ILLIQ去掉nan
    tmp_factor = factor_pure.ix[dt].dropna()
    tmp_return = forward_20d_return_data.ix[dt].dropna()
    tmp_return = tmp_return[tmp_return<0.6]
    tmp_return_mean = tmp_return.mean()
    
    pct_quantiles = 1.0/n_quantile
    for i in range(n_quantile):
        down = tmp_factor.quantile(pct_quantiles*i)
        up = tmp_factor.quantile(pct_quantiles*(i+1))
        i_quantile_index = tmp_factor[(tmp_factor<=up) & (tmp_factor>=down)].index
        mean_tmp = tmp_return[i_quantile_index].mean() - tmp_return_mean
        qt_mean_results.append(mean_tmp)
        
    excess_returns_means.ix[dt] = qt_mean_results

excess_returns_means.dropna(inplace=True)

# 因子分组的超额收益作图
fig = plt.figure(figsize=(12, 6))
ax1 = fig.add_subplot(111)

excess_returns_means_dist = excess_returns_means.mean()
excess_dist_plus = excess_returns_means_dist[excess_returns_means_dist>0]
excess_dist_minus = excess_returns_means_dist[excess_returns_means_dist<0]
lns2 = ax1.bar(excess_dist_plus.index, excess_dist_plus.values, align='center', color='r', width=0.35)
lns3 = ax1.bar(excess_dist_minus.index, excess_dist_minus.values, align='center', color='g', width=0.35)

ax1.set_xlim(left=0.5, right=len(excess_returns_means_dist)+0.5)
# ax1.set_ylim(-0.008, 0.008)
ax1.set_ylabel(u'超额收益', fontproperties=font, fontsize=16)
ax1.set_xlabel(u'十分位分组', fontproperties=font, fontsize=16)
ax1.set_xticks(excess_returns_means_dist.index)
ax1.set_xticklabels([int(x) for x in ax1.get_xticks()], fontproperties=font, fontsize=14)
ax1.set_yticklabels([str(x*100)+'0%' for x in ax1.get_yticks()], fontproperties=font, fontsize=14)
ax1.set_title(u"因子超额收益", fontproperties=font, fontsize=16)
ax1.grid()
```

可以看到，该因子选股不同分位数组合的超额收益呈很好的单调性，说明因子具有良好的区分能力

```python
factor_rev = pd.DataFrame(columns=factor.columns, index=factor.index)
for date in factor.index:
    current_universe = factor.loc[date, :].dropna().index & backward_20d_return_data.loc[date, :].dropna().index
    x = backward_20d_return_data.loc[date, current_universe]
    y = factor.loc[date, current_universe]
    x = np.array(x).reshape((-1,1))
    y = np.array(y)
    linreg = LinearRegression()
    model = linreg.fit(x, y)
    y_pre = model.predict(x)
    residual = y - y_pre
    factor_rev.ix[date, current_universe] = residual
```
