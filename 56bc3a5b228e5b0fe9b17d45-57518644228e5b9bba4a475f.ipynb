# 双均线战法 5日均线和60日均线

双均线策略：  
当5日均线上穿60日均线，买入  
当5日均线下穿60日均线，卖出  

为什么选MA5和MA60？因为我简单实验了下，比如(5,10),(5,20)，(5,60)，(10,20)，(10,60),(20,60)。。，甚至《海龟交易法则》提到的（100,350），针对指数ETF(5,60)效果最好。
