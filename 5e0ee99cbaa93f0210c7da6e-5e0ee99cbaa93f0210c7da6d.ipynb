## 个股公告

获取日历

```python
import numpy as np
import pandas as pd
from CAL.PyCAL import *
from datetime import date

cal = Calendar('China.SSE')

def myTradeDate(predays="-1B",fmt="%Y%m%d"):
    """
    获取交易日期
    eg. 
    # 获取倒数第20个交易日
    # myTradeDate("-20B")
    """
    return cal.advanceDate(date.today(), predays,  BizDayConvention.Following).strftime(fmt)

# 开始时间为1个月（20个交易日）
beginDate=myTradeDate('-20B')

ticker="603799"
```

根据证券代码获取相应公告分类结果

```python
DataAPI.ReportByTickerGet(ticker=ticker,beginDate=beginDate,endDate=u"",field=u"",pandas="1")
```
