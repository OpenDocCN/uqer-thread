# 2015年国科大量化投资课程期末总结—动量策略

　　动量效应是指在一定持有期内，如果某只股票或者某个股票组合在前一段时期表现较好，那么，下一段时期该股票或者股票投资组合仍将有良好表现。从熊市来看（2008 年-2013 年） ，股票市场存在较强的动量效应。

# 策略计算数据与流程
** 选择样本: ** 2008年-2013年沪深300（这些股票表征了大盘的整体走势，便于核查策略相对于大盘整体的好坏。）

** 选取赢家组合的标准：**  前3（6, 12）月个股收益率 mrt3（mrt6, mrt12）

** 持有期：** 3（6, 12）个月

** 单边成本：**0.1%

** 流程：**

1. 在构建投资组合的期间，根据 mt3（mt6, mrt12）从大到小依次对样本股票排序；
2. 选取前10%的股票，将现有资金按照选取股票的数量平分，分别购入这些股票； 
3. 持有期分别为 3（6, 12）个月，在每个持有期末卖出已不属于10%的股票，将卖出所得资金重新平分买入新加入前10% 的股票；
4. 在持有期末，重新构建组合，即重复第二至第三步的操作。

# 程序实现

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

universe = set_universe('HS300')

Data_date = DataAPI.TradeCalGet(exchangeCD="XSHG",beginDate="20080101",endDate="20131231",field=["calendarDate","isMonthEnd"],pandas="1")
calendarDate = Data_date.calendarDate.values
isMonthEnd = Data_date.isMonthEnd.values
tradeDate = []
for i in range(len(calendarDate)):
    if isMonthEnd[i] == 1:
        calendar = calendarDate[i].replace("-","")
        tradeDate.append(calendar)                     # 所有交易日期

def TimeCollect(Freq, tradeDate, mrt):                 # 形成期和持有期的节点
    TimeI = []                                         # 建仓平仓点
    TimeII = []                                        # 选股收益起始点
    n = (len(tradeDate) - mrt)/Freq
    for i in range(n):
        TimeI.append(tradeDate[int(Freq * i) + mrt])
        TimeII.append(tradeDate[int(Freq * i)])
    return TimeI, TimeII

def FreqBeneMonthly(ticker_code, timeXC, time):        # 选股排序收益率
    try: 
        temp1 = DataAPI.MktEqudGet(secID=ticker_code, tradeDate=timeXC, field=['closePrice'], pandas="1")
        temp2 = DataAPI.MktEqudGet(secID=ticker_code, tradeDate=time, field=['closePrice'], pandas="1")
    except:
        temp1 = pd.DataFrame()
        temp2 = pd.DataFrame()
    #print temp1
    if len(temp1) > 0:
        benefit = np.log(temp1.closePrice)-np.log(temp2.closePrice)
        return benefit
    else:
        return np.nan

def SignalSet(time1, time2):                          # 选取前10%排序
    temp = pd.Series(index=universe)
    for each in temp.index:
        temp[each] = FreqBeneMonthly(each, time1, time2)
    #print temp
    quan_90 = temp.quantile(0.9)
    selectList = list(temp.index[temp >= quan_90])
    return selectList

def ChongFu(List1, List2):                             # 数列重复
    ListChongFu = []
    ListShengXia1 = []
    ListShengXia2 = []
    MarkChongFu1 = []
    MarkChongFu2 = []
    MarkShengXia1 = []
    MarkShengXia2 = []
    for i in range(len(List1)):
        mark1 = -1
        for j in range(len(List2)):
            if List1[i] == List2[j]:
                mark1 = i
        if i != mark1:
            ListShengXia1.append(List1[i])
            MarkShengXia1.append(i)
        else:
            ListChongFu.append(List1[i])
            MarkChongFu1.append(i)
    for i in range(len(List2)):
        mark2 = -1
        for j in range(len(List1)):
            if List2[i] == List1[j]:
                mark2 = i
        if i != mark2:
            ListShengXia2.append(List2[i])
            MarkShengXia2.append(i)
        else:
            MarkChongFu2.append(i)
    if len(ListChongFu) == 0:
        Judge = False
    else:
        Judge = True
    return Judge, ListChongFu, ListShengXia1, ListShengXia2, MarkChongFu1,MarkChongFu2, MarkShengXia1, MarkShengXia2         

Mrt = [3,6,12]
Freq = [3,6,12] #3,6,12

BenefitSing = [[] for i in range(len(Mrt))]
BenefitAccu = [[] for i in range(len(Mrt))]
Suc = [[] for i in range(len(Freq))]
Shaper = [[] for i in range(len(Freq))]
BoDong = [[] for i in range(len(Freq))]
NianHuaShouYi = [[] for i in range(len(Freq))]
MaxDrawBack = [[] for i in range(len(Freq))]


for f in range(len(Freq)):
    BenefitSing = [[] for i in range(len(Mrt))]
    BenefitAccu = [[] for i in range(len(Mrt))]
    print("持有期：%d"%Freq[f] + r"个月")
    for M in range(len(Mrt)):
        mrt = Mrt[M]
        TimeXC, Time =  TimeCollect(Freq[f], tradeDate, mrt)
        amount = []
        Price = 1000000
        ZuiGao = Price
        maxDrawBack = 0
        ChengBen = 0
        countTot = 0
        countSuc = 0
        print("形成期：%d"%Mrt[M] + r"个月")
        for t in range(len(TimeXC)):
            prePrice = Price
            if len(amount) == 0:
                selectList = SignalSet(TimeXC[t],Time[t])
                price0 = Price/len(selectList)
                for i in range(len(selectList)):
                    data_GuPiao = DataAPI.MktEqudGet(secID=selectList[i], tradeDate=TimeXC[t], field=['closePrice'], pandas="1")
                    amount.append(price0/data_GuPiao.closePrice.values[0])
                ChengBen = Price * 0.001
                selectList2 = selectList
            else:
                price1 = 0
                Price = 0
                selectList1 = SignalSet(TimeXC[t],Time[t])
                judge, ListChongFu, ListShengXia2, ListShengXia1, MarkChongFu2,MarkChongFu1, MarkShengXia2, MarkShengXia1  = ChongFu(selectList2, selectList1)
                if judge == True:
                    selectList = []
                    amount0 = []
                    for i in range(len(ListShengXia2)):
                        data_GuPiao = DataAPI.MktEqudGet(secID=ListShengXia2[i], tradeDate=TimeXC[t], field=['closePrice'], pandas="1")
                        Price = Price + amount[MarkShengXia2[i]] * data_GuPiao.closePrice.values[0] * (1-0.001)
                    for i in range(len(ListChongFu)):
                        data_GuPiao = DataAPI.MktEqudGet(secID=ListChongFu[i], tradeDate=TimeXC[t], field=['closePrice'], pandas="1")
                        selectList.append(ListChongFu[i])
                        amount0.append(amount[MarkChongFu2[i]])
                        price1 = price1 + amount[MarkChongFu2[i]]*data_GuPiao.closePrice.values[0]
                    Price = Price - ChengBen
                    price0 = Price/len(ListShengXia1)
                    for i in ListShengXia1:
                        data_GuPiao = DataAPI.MktEqudGet(secID=i, tradeDate=TimeXC[t], field=['closePrice'], pandas="1")
                        selectList.append(i)
                        amount0.append(price0/data_GuPiao.closePrice.values[0])
                    ChengBen = Price * 0.001
                    Price = Price + price1
                else:
                    selectList = []
                    amount0 = []
                    for i in range(len(selectList2)):
                        data_GuPiao = DataAPI.MktEqudGet(secID=selectList2[i], tradeDate=TimeXC[t], field=['closePrice'], pandas="1")
                        Price = Price + amount[i] * data_GuPiao.closePrice.values[0] * (1-0.001)
                    Price = Price - ChengBen
                    price0 = Price/len(selectList1)
                    for i in range(len(selectList1)):
                        data_GuPiao = DataAPI.MktEqudGet(secID=selectList1[i], tradeDate=TimeXC[t], field=['closePrice'], pandas="1")
                        selectList.append(selectList1[i])
                        amount0.append(price0/data_GuPiao.closePrice.values[0])
                    ChengBen = Price * 0.001
                selectList2 = selectList
                amount = amount0
                countTot = countTot + 1
                BenefitSing[M].append((Price - prePrice)/prePrice)
                BenefitAccu[M].append((Price-1000000)/1000000)
                if Price >= prePrice:
                    countSuc = countSuc + 1
                else:
                    countSuc = countSuc
                ZuiGao = max(ZuiGao, Price)
                maxDrawBack = max(1 - Price/ZuiGao, maxDrawBack)
        Suc[f].append(float(countSuc)/countTot)
        MaxDrawBack[f].append(maxDrawBack)
        BoDong[f].append(np.std(BenefitSing[M])/Freq[f]*12)
        Shaper[f].append((np.average(BenefitSing[M]) - 0.0238/12*Freq[f])/np.std(BenefitSing[M]))
        NianHuaShouYi[f].append(BenefitAccu[M][-1]/Freq[f]*12)
    
    
    
        print('年化收益率：%.2f'%(100*NianHuaShouYi[f][M]) + r'%')
        print('夏普比率：%.2f'%(Shaper[f][M]))
        print('波动率：%.2f'%(100*BoDong[f][M]) + r'%')
        print('最大回撤：%.2f'%(100*MaxDrawBack[f][M]) + r'%')
        print('成功率：%.2f'%(100*Suc[f][M]) + r'%')
        print ""

    colors = ["blue","red","green"]
    plt.subplot(211)
    for i in range(3):
        plt.plot(BenefitSing[i],color=colors[i])
    plt.title("Single")
    plt.xlabel("Date")
    plt.ylabel("Price")

    plt.subplot(212)
    for i in range(3):
        plt.plot(BenefitAccu[i],color=colors[i])
    plt.title("Accumulate")
    plt.xlabel("Date")
    plt.ylabel("Price")
    plt.show() 
    
    print ""
    print ""
```

# 程序拓展

　　在原始程序中，仅对HS300取不同的（持有期，形成期）组合，进行了横向纵向对比，并没有对大盘走势有详细对比，本程序使用了HS300从2008年到2013年每月末的指数信息，添加了新函数，用以计算对应不同（持有期，形成期）组合下，持有期前后的HS300指数变化（年化收益率、夏普比率、波动率、最大回撤、成功率）。
        
　　说明：在此程序仅列出HS300数据导入、新增加函数和改动后主程序。
 
　　　　　程序仅以（持有期，形成期）为（3，3）为例。

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

HS300 = pd.read_excel("HS300-week.xlsx")
closePriceHS = HS300["closePrice"].values

universe = set_universe('HS300')

Data_date = DataAPI.TradeCalGet(exchangeCD="XSHG",beginDate="20080101",endDate="20131231",field=["calendarDate","isMonthEnd"],pandas="1")
calendarDate = Data_date.calendarDate.values
isMonthEnd = Data_date.isMonthEnd.values
tradeDate = []
for i in range(len(calendarDate)):
    if isMonthEnd[i] == 1:
        calendar = calendarDate[i].replace("-","")
        tradeDate.append(calendar)                     # 所有交易日期

def TimePriceCollect(Freq, tradeDate, closePrice, mrt):                 # 形成期和持有期的节点,HS300持有期收盘价
    TimeI = []                                                          # 建仓平仓点
    TimeII = []                                                         # 选股收益起始点
    closePriceNew = []
    n = (len(tradeDate) - mrt)/Freq
    for i in range(n):
        TimeI.append(tradeDate[int(Freq * i) + mrt])
        closePriceNew.append(closePrice[int(Freq * i) + mrt])
        TimeII.append(tradeDate[int(Freq * i)])
    return TimeI, TimeII, closePriceNew

def FreqBeneMonthly(ticker_code, timeXC, time):        # 选股排序收益率
    try: 
        temp1 = DataAPI.MktEqudGet(secID=ticker_code, tradeDate=timeXC, field=['closePrice'], pandas="1")
        temp2 = DataAPI.MktEqudGet(secID=ticker_code, tradeDate=time, field=['closePrice'], pandas="1")
    except:
        temp1 = pd.DataFrame()
        temp2 = pd.DataFrame()
    #print temp1
    if len(temp1) > 0:
        benefit = np.log(temp1.closePrice)-np.log(temp2.closePrice)
        return benefit
    else:
        return np.nan

def SignalSet(time1, time2):                          # 选取前10%排序
    temp = pd.Series(index=universe)
    for each in temp.index:
        temp[each] = FreqBeneMonthly(each, time1, time2)
    #print temp
    quan_90 = temp.quantile(0.9)
    selectList = list(temp.index[temp >= quan_90])
    return selectList

def ChongFu(List1, List2):                             # 数列重复
    ListChongFu = []
    ListShengXia1 = []
    ListShengXia2 = []
    MarkChongFu1 = []
    MarkChongFu2 = []
    MarkShengXia1 = []
    MarkShengXia2 = []
    for i in range(len(List1)):
        mark1 = -1
        for j in range(len(List2)):
            if List1[i] == List2[j]:
                mark1 = i
        if i != mark1:
            ListShengXia1.append(List1[i])
            MarkShengXia1.append(i)
        else:
            ListChongFu.append(List1[i])
            MarkChongFu1.append(i)
    for i in range(len(List2)):
        mark2 = -1
        for j in range(len(List1)):
            if List2[i] == List1[j]:
                mark2 = i
        if i != mark2:
            ListShengXia2.append(List2[i])
            MarkShengXia2.append(i)
        else:
            MarkChongFu2.append(i)
    if len(ListChongFu) == 0:
        Judge = False
    else:
        Judge = True
    return Judge, ListChongFu, ListShengXia1, ListShengXia2, MarkChongFu1,MarkChongFu2, MarkShengXia1, MarkShengXia2

BenefitSing = [[],[]]
BenefitAccu = [[],[]]

Suc = []
MaxDrawBack = []
BoDong = []
Shaper = []
NianHuaShouYi = []

Freq = 3
mrt = 3
TimeXC, Time, closePriceHS300 =  TimePriceCollect(Freq, tradeDate, closePriceHS, mrt)
amount = []
Price = 1000000
ZuiGao = Price
maxDrawBack = 0
ChengBen = 0
countTot = 0
countSuc = 0
for t in range(len(TimeXC)):
    prePrice = Price
    if len(amount) == 0:
        selectList = SignalSet(TimeXC[t],Time[t])
        price0 = Price/len(selectList)
        for i in range(len(selectList)):
            data_GuPiao = DataAPI.MktEqudGet(secID=selectList[i], tradeDate=TimeXC[t], field=['closePrice'], pandas="1")
            amount.append(price0/data_GuPiao.closePrice.values[0])
        ChengBen = Price * 0.001
        selectList2 = selectList
    else:
        price1 = 0
        Price = 0
        selectList1 = SignalSet(TimeXC[t],Time[t])
        judge, ListChongFu, ListShengXia2, ListShengXia1, MarkChongFu2,MarkChongFu1, MarkShengXia2, MarkShengXia1  = ChongFu(selectList2, selectList1)
        if judge == True:
            selectList = []
            amount0 = []
            for i in range(len(ListShengXia2)):
                data_GuPiao = DataAPI.MktEqudGet(secID=ListShengXia2[i], tradeDate=TimeXC[t], field=['closePrice'], pandas="1")
                Price = Price + amount[MarkShengXia2[i]] * data_GuPiao.closePrice.values[0] * (1-0.001)
            for i in range(len(ListChongFu)):
                data_GuPiao = DataAPI.MktEqudGet(secID=ListChongFu[i], tradeDate=TimeXC[t], field=['closePrice'], pandas="1")
                selectList.append(ListChongFu[i])
                amount0.append(amount[MarkChongFu2[i]])
                price1 = price1 + amount[MarkChongFu2[i]]*data_GuPiao.closePrice.values[0]
            Price = Price - ChengBen
            price0 = Price/len(ListShengXia1)
            for i in ListShengXia1:
                data_GuPiao = DataAPI.MktEqudGet(secID=i, tradeDate=TimeXC[t], field=['closePrice'], pandas="1")
                selectList.append(i)
                amount0.append(price0/data_GuPiao.closePrice.values[0])
            ChengBen = Price * 0.001
            Price = Price + price1
        else:
            selectList = []
            amount0 = []
            for i in range(len(selectList2)):
                data_GuPiao = DataAPI.MktEqudGet(secID=selectList2[i], tradeDate=TimeXC[t], field=['closePrice'], pandas="1")
                Price = Price + amount[i] * data_GuPiao.closePrice.values[0] * (1-0.001)
            Price = Price - ChengBen
            price0 = Price/len(selectList1)
            for i in range(len(selectList1)):
                data_GuPiao = DataAPI.MktEqudGet(secID=selectList1[i], tradeDate=TimeXC[t], field=['closePrice'], pandas="1")
                selectList.append(selectList1[i])
                amount0.append(price0/data_GuPiao.closePrice.values[0])
            ChengBen = Price * 0.001
        selectList2 = selectList
        amount = amount0
        countTot = countTot + 1
        BenefitSing[0].append((Price - prePrice)/prePrice)
        BenefitAccu[0].append((Price-1000000)/1000000)
        if Price >= prePrice:
            countSuc = countSuc + 1
        else:
            countSuc = countSuc
        ZuiGao = max(ZuiGao, Price)
        maxDrawBack = max(1 - Price/ZuiGao, maxDrawBack)
Suc.append(float(countSuc)/countTot)
MaxDrawBack.append(maxDrawBack)
BoDong.append(np.std(BenefitSing[0])/Freq*12)
Shaper.append((np.average(BenefitSing[0]) - 0.0238/12*Freq)/np.std(BenefitSing[0]))
NianHuaShouYi.append(BenefitAccu[0][-1]/Freq*12)

Price = 1000000
ZuiGao = Price
maxDrawBack = 0
ChengBen = 0
countTot = 0
countSuc = 0
amount1 = Price/closePriceHS300[0]
for t in range(1,len(TimeXC)):
    prePrice = Price
    Price = amount1 * closePriceHS300[t]
    BenefitSing[1].append((Price - prePrice)/prePrice)
    BenefitAccu[1].append((Price-1000000)/1000000)
    countTot = countTot + 1
    if Price >= prePrice:
        countSuc = countSuc + 1
    else:
        countSuc = countSuc
    ZuiGao = max(ZuiGao, Price)
    maxDrawBack = max(1 - Price/ZuiGao, maxDrawBack)

Suc.append(float(countSuc)/countTot)
MaxDrawBack.append(maxDrawBack)
BoDong.append(np.std(BenefitSing[1])/Freq*12)
Shaper.append((np.average(BenefitSing[1]) - 0.0238/12*Freq)/np.std(BenefitSing[1]))
NianHuaShouYi.append(BenefitAccu[1][-1]/Freq*12)

print("持有期：%d"%Freq + r"个月")
print("形成期：%d"%mrt + r"个月")
print("")
print("策略情况")
print('年化收益率：%.2f'%(100*NianHuaShouYi[0]) + r'%')
print('夏普比率：%.2f'%(Shaper[0]))
print('波动率：%.2f'%(100*BoDong[0]) + r'%')
print('最大回撤：%.2f'%(100*MaxDrawBack[0]) + r'%')
print('成功率：%.2f'%(100*Suc[0]) + r'%')
print("")
print("HS300情况")
print('年化收益率：%.2f'%(100*NianHuaShouYi[1]) + r'%')
print('夏普比率：%.2f'%(Shaper[1]))
print('波动率：%.2f'%(100*BoDong[1]) + r'%')
print('最大回撤：%.2f'%(100*MaxDrawBack[1]) + r'%')
print('成功率：%.2f'%(100*Suc[1]) + r'%')


colors = ["blue","red"]
plt.subplot(211)
for i in range(2):
    plt.plot(BenefitSing[i],color=colors[i])
plt.title("Single")
plt.xlabel("Date")
plt.ylabel("Price")

plt.subplot(212)
for i in range(2):
    plt.plot(BenefitAccu[i],color=colors[i])
plt.title("Accumulate")
plt.xlabel("Date")
plt.ylabel("Price")
plt.show() 
```
