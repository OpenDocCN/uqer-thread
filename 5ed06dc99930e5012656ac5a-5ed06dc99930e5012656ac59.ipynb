```python
import pandas as pd
import pandas.io.data as web
import matplotlib.pyplot as plt
import datetime
import numpy as np
from dateutil.relativedelta import relativedelta


def get_draw_down(data):
    #为了得到错位向量，构造新向量
    original = np.array(data)
    # print(original)
    minuend = np.append(original, 0)
    subtractor = np.insert(original,0, 0)
    
    #得到错位向量diff
    diffs = minuend - subtractor
    #头尾值无效，无参考性
    diffs = diffs[1:-1]
    
    #计算历次回撤率
    ends=[] #用于显示图，记录局部最低点的日期
    values = [] #储存历次回撤率
    start=None #记录局部最大值的index
    gap = 0 #记录gap值
    is_first = True #正负数交界开关
    for index, diff in enumerate(diffs):
        if diff < 0:
            gap += diff
            if is_first is True:
                start = index
                is_first = False

        else:
            if is_first is False:
                values.append(abs(gap) / original[start])
                ends.append(index)
                # print("gap is {}".format(gap))
                # print("max is {}".format(original[start]))
                is_first = True
            gap = 0

    if is_first is False:
        # print("gap is {}".format(gap))
        # print("max is {}".format(original[start]))
        values.append(abs(gap) / original[start])
        
    return values,ends
    

#设置危机起始日期        
start = datetime.datetime(2007, 4, 1)
end = datetime.datetime(2008, 8, 1)

#获得数据
out_data = DataAPI.MktEqudGet(secID=u"000001.XSHE", beginDate=start, endDate=end, pandas="1")

#计算历次回撤率
data,indecies=get_draw_down(list(out_data['preClosePrice']))

#绘图
labels=[out_data['tradeDate'][index] for index in indecies]

plt.figure(figsize=(18,4))
plt.xticks([i for i in range(len(labels))],labels,rotation=60)
plt.title('Draw Down Analysis')
plt.xlabel("date")
plt.ylabel("draw down rate")
plt.plot([i for i in range(len(data))],data)

plt.show()
```
