## 看一下每日股票的变化情况：

```python
dailysmallstock = [sorted(x.keys()) for x in bt.security_position]
result = []
for i in range(len(dailysmallstock)):
    if i == 0:
        continue
    tmpdata = list(set(dailysmallstock[i]).difference(set(dailysmallstock[i-1])))
    if len(tmpdata) != 0:
        #print '[{0}]个交易日股票有变动:\n{1}'.format(i, ','.join(tmpdata))
        result.append(len(tmpdata)/10.)

print '日均变动率：',sum(result)/len(result)
```
