# 传统技术指标结合机器学习发掘超额收益组合

**思路：量价信息形态信息为特征，使用收益率作为预测目标（未来10，20，40，60日最高收益，收盘收益等，下周期最高收益，下周期收盘收益）**
**举例：**
1. closeIndex_ema5_derivatives：	5日加权ema的方向导数，含有正负取值信息
5. DIF_DEA_relative_derivatives：macd的dif和dea相对位置，(dif-dea).diff()
等等。

**注意事项**
1，将技术指标信息转化为连续值，通过导数正负和幅度确定技术特征
2，相对位置信息依然采用-1和1的离散取值表示
3，预测目标为未来5日内的最高价比率，大于等于5%认为1，否则0（非5日后价格，个人感觉这个较为容易，可与止盈策略搭配使用）


```python
import CAL.PyCAL as pycal
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from quartz.api import set_universe
import talib 

import seaborn as sns
sns.set_style('white')
```

```python
from sklearn.preprocessing import StandardScaler

#参数赋值
start='20140106'
end='20180915'
#样例指数
exampleTicker='399006'
#均线等使用等窗口参数
periodWindow=10

marketIndexDataAll=pd.DataFrame();

#当前所有数据维度
tickerAll=DataAPI.MktIdxdGet(indexID=u"",ticker=u"",tradeDate=u"20140106",beginDate=u"",endDate=u"",exchangeCD=u"XSHE,XSHG",field=['ticker'],pandas="1")
for exampleTickerList in tickerAll.values.tolist()[0:10]:
    exampleTicker=exampleTickerList[0]

    #获取指数行情数据
    marketIndexData=DataAPI.MktIdxdGet(indexID=u"",ticker=exampleTicker,tradeDate='',beginDate=start,endDate=end,field=['tradeDate','closeIndex','turnoverVol','lowestIndex','highestIndex'],pandas="1")
    #列tradeDate修改为日期类型并且设置为索引
    marketIndexData['tradeDate'] = pd.to_datetime(marketIndexData['tradeDate'])
    marketIndexData.set_index("tradeDate", inplace=True)

    ###添加特征，均线相关
    #加权移动平均,导数
    marketIndexData['closeIndex_ema5']=pd.ewma(marketIndexData['closeIndex'], span=5)
    marketIndexData['closeIndex_ema5_derivatives']=marketIndexData['closeIndex_ema5'].diff().fillna(0)/marketIndexData['closeIndex_ema5']
    marketIndexData['closeIndex_ema5_derivatives']=StandardScaler().fit_transform(marketIndexData['closeIndex_ema5_derivatives'].values)

    marketIndexData['closeIndex_ema10']=pd.ewma(marketIndexData['closeIndex'], span=10)
    marketIndexData['closeIndex_ema10_derivatives']=marketIndexData['closeIndex_ema10'].diff().fillna(0)/marketIndexData['closeIndex_ema10']
    marketIndexData['closeIndex_ema10_derivatives']=StandardScaler().fit_transform(marketIndexData['closeIndex_ema10_derivatives'].values)

    marketIndexData['closeIndex_ema20']=pd.ewma(marketIndexData['closeIndex'], span=20)
    marketIndexData['closeIndex_ema20_derivatives']=marketIndexData['closeIndex_ema20'].diff().fillna(0)/marketIndexData['closeIndex_ema20']
    marketIndexData['closeIndex_ema20_derivatives']=StandardScaler().fit_transform(marketIndexData['closeIndex_ema20_derivatives'].values)

    marketIndexData['closeIndex_ema60']=pd.ewma(marketIndexData['closeIndex'], span=60)
    marketIndexData['closeIndex_ema60_derivatives']=marketIndexData['closeIndex_ema60'].diff().fillna(0)/marketIndexData['closeIndex_ema60']
    marketIndexData['closeIndex_ema60_derivatives']=StandardScaler().fit_transform(marketIndexData['closeIndex_ema60_derivatives'].values)

    
    #均线相对位置差异,0-5.5-10,5-20,10-20,相对位置和diff的导数
    marketIndexData['closeIndex_ema0_5']=(marketIndexData['closeIndex']-marketIndexData['closeIndex_ema5'])
    marketIndexData['closeIndex_ema0_5_derivatives']= marketIndexData['closeIndex_ema0_5'].diff().fillna(0)
    marketIndexData['closeIndex_ema0_5_derivatives']=StandardScaler().fit_transform(marketIndexData['closeIndex_ema0_5_derivatives'].values)
    marketIndexData['closeIndex_ema0_5']=marketIndexData['closeIndex_ema0_5'].apply(lambda x:1 if x>0 else -1)
    
    marketIndexData['closeIndex_ema5_10']=marketIndexData['closeIndex_ema5']-marketIndexData['closeIndex_ema10']
    marketIndexData['closeIndex_ema5_10_derivatives']=marketIndexData['closeIndex_ema5_10'].diff().fillna(0)
    marketIndexData['closeIndex_ema5_10_derivatives']=StandardScaler().fit_transform(marketIndexData['closeIndex_ema5_10_derivatives'].values)
    marketIndexData['closeIndex_ema5_10']=marketIndexData['closeIndex_ema5_10'].apply(lambda x:1 if x>0 else -1)
    
    marketIndexData['closeIndex_ema10_20']=marketIndexData['closeIndex_ema10']-marketIndexData['closeIndex_ema20']
    marketIndexData['closeIndex_ema10_20_derivatives']=marketIndexData['closeIndex_ema10_20'].diff().fillna(0)
    marketIndexData['closeIndex_ema10_20_derivatives']=StandardScaler().fit_transform(marketIndexData['closeIndex_ema10_20_derivatives'].values)
    marketIndexData['closeIndex_ema10_20']=marketIndexData['closeIndex_ema10_20'].apply(lambda x:1 if x>0 else -1)

    ###预测信息
    #未来5，10，20日最高价，收盘价,TODO EMA的最高价
    marketIndexData['closeIndex_max5_future']=((marketIndexData['closeIndex'].rolling(5).max().shift(-5)-marketIndexData['closeIndex'])/marketIndexData['closeIndex']).apply(lambda x:int(x*100) if not np.isnan(x) else 0)
    marketIndexData['closeIndex_close5_future']=((marketIndexData['closeIndex'].shift(-5)-marketIndexData['closeIndex'])/marketIndexData['closeIndex']).apply(lambda x:int(x*100) if not np.isnan(x) else 0)

    #添加macd线信息
    #2种信息。
    # 1，dif和dea的相对位置。
    # 2, dif和dea的交叉信息，上穿下-2，下穿上2。
    DIF,DEA,MACD = talib.MACD(marketIndexData['closeIndex'].values, fastperiod=12, slowperiod=26, signalperiod=20)
    marketIndexData['DIF_DEA_relative']=DIF-DEA
    marketIndexData['DIF_DEA_relative_derivatives']=marketIndexData['DIF_DEA_relative'].diff().fillna(0)
    marketIndexData['DIF_DEA_relative_derivatives']=StandardScaler().fit_transform(marketIndexData['DIF_DEA_relative_derivatives'].values)
    marketIndexData['DIF_DEA_relative']=marketIndexData['DIF_DEA_relative'].apply(lambda x:1 if x>0 else -1)

    #添加kdj信息
    # 1，k线和d线相对位置
    # 2，d线大于80
    # 3，d线小于20
    k,d=talib.STOCH(np.array(marketIndexData['highestIndex']),np.array(marketIndexData['lowestIndex']),np.array(marketIndexData['closeIndex']),
      fastk_period=9,slowk_period=3,slowk_matype=0,slowd_period=3,slowd_matype=0)
    marketIndexData['K_D_relative']=k-d
    marketIndexData['K_D_relative_derivatives']=marketIndexData['K_D_relative'].diff().fillna(0)
    marketIndexData['K_D_relative_derivatives']=StandardScaler().fit_transform(marketIndexData['K_D_relative_derivatives'].values)
    marketIndexData['K_D_relative']=marketIndexData['K_D_relative'].apply(lambda x:1 if x>0 else -1)
 
    #添加RSA信息
    #1，大于80
    #2，小于20
    rsi=talib.RSI(np.array(marketIndexData['closeIndex']), timeperiod=12)  
    marketIndexData['RSA_80_relative']=rsi
    marketIndexData['RSA_80_relative']=marketIndexData['RSA_80_relative'].apply(lambda x:1 if x>=80 else 0).fillna(0)
    marketIndexData['RSA_80_relative']=StandardScaler().fit_transform(marketIndexData['RSA_80_relative'].values)
    marketIndexData['RSA_20_relative']=rsi
    marketIndexData['RSA_20_relative']=marketIndexData['RSA_20_relative'].apply(lambda x:1 if x<=20 else 0).fillna(0)
    marketIndexData['RSA_20_relative']=StandardScaler().fit_transform(marketIndexData['RSA_20_relative'].values)
    
  
    marketIndexData_2=marketIndexData[[ 'closeIndex_max5_future',
                                        'closeIndex_close5_future',
                                        # 'closeIndex_ema5',
                                        'closeIndex_ema5_derivatives',
                                        # 'closeIndex_ema10',
                                        'closeIndex_ema10_derivatives',
                                        # 'closeIndex_ema20',
                                        'closeIndex_ema20_derivatives',
                                        # 'closeIndex_ema60',
                                        'closeIndex_ema60_derivatives',
                                        'closeIndex_ema0_5',
                                        'closeIndex_ema0_5_derivatives',
                                        'closeIndex_ema5_10',
                                        'closeIndex_ema5_10_derivatives',
                                        'closeIndex_ema10_20',
                                        'closeIndex_ema10_20_derivatives',
                                        'DIF_DEA_relative',
                                        'DIF_DEA_relative_derivatives',
                                        'K_D_relative',
                                        'K_D_relative_derivatives',
                                        'RSA_80_relative',
                                        'RSA_20_relative'
                                 ]]
    marketIndexDataAll = pd.concat([marketIndexDataAll, marketIndexData_2], axis=0)


    
print marketIndexDataAll.columns
```
