# 价差扩大 

本来想通过研究价差的统计关系，看跨期价差交易在实践中的运用，而在测试中发现，在很少的极端行情中，价差存在大幅度扩大的可能

```python
## 豆粕期货 1，5，9 合约数据采选 ##
# 初始环境及信息准备 #

import pandas as pd 
import DataAPI as DA
import numpy as np
import matplotlib as mp
import matplotlib.pyplot as plt
import datetime
import math

startdate='2011-01-01'
enddate='2018-08-12'
fee=6                     # 单边手续费
value=1000000
margin=0.05

m1=[]
m5=[]
m9=[]

# 准备合约代码 #
for i in range(1,10):
    m1.append('M'+str(i*100+1+1000))
    m5.append('M'+str(i*100+5+1000))
    m9.append('M'+str(i*100+9+1000))

# 提取交易日 #
tradedate=DA.TradeCalGet(exchangeCD=u"XDCE",beginDate=startdate,endDate=enddate,field=u"",pandas="1")
td= tradedate[tradedate['isOpen']==1]
td=td[u'calendarDate']
td=td.reset_index()[u'calendarDate']
dta1=pd.DataFrame([])
dta5=pd.DataFrame([])
dta9=pd.DataFrame([])
for j in range(1,10):
# 提取1月合约数据 #
    dt1=[dta1,pd.DataFrame(DataAPI.MktFutdVolGet(secID=u"",ticker=m1[j-1],beginDate=startdate,endDate=enddate,field=u"",pandas="1"))]
    dta1=pd.concat(dt1)
    dta1=dta1.reset_index(drop=True)
# 提取5月合约数据 #
    dt5=[dta5,pd.DataFrame(DataAPI.MktFutdVolGet(secID=u"",ticker=m5[j-1],beginDate=startdate,endDate=enddate,field=u"",pandas="1"))]
    dta5=pd.concat(dt5)
    dta5=dta5.reset_index(drop=True)
# 提取9月合约数据 #
    dt9=[dta9,pd.DataFrame(DataAPI.MktFutdVolGet(secID=u"",ticker=m9[j-1],beginDate=startdate,endDate=enddate,field=u"",pandas="1"))]
    dta9=pd.concat(dt9)
    dta9=dta9.reset_index(drop=True)
dta1=dta1[[1,3,4,9,10,11,12,13,14,15,16]]
dta5=dta5[[1,3,4,9,10,11,12,13,14,15,16]]
dta9=dta9[[1,3,4,9,10,11,12,13,14,15,16]]
```

```python
dta1
```

```python
# 通过分析成交，构建远近月合约

vv1=dta1['openInt']
vv5=dta5['openInt']
vv9=dta9['openInt']
mc=pd.DataFrame()
mc2=pd.DataFrame()
for s in range (1,1851):
    if vv1[s-1]>=vv9[s-1] and vv5[s-1]>=vv9[s-1]:
        mcc=[mc,pd.DataFrame(dta1.values[s-1]).T]
        mc=pd.concat(mcc)
        mcc2=[mc2,pd.DataFrame(dta5.values[s-1]).T]
        mc2=pd.concat(mcc2)
    elif vv5[s-1]>=vv1[s-1] and vv9[s-1]>=vv1[s-1]:
        mcc=[mc,pd.DataFrame(dta5.values[s-1]).T]
        mc=pd.concat(mcc)
        mcc2=[mc2,pd.DataFrame(dta9.values[s-1]).T]
        mc2=pd.concat(mcc2)
    elif vv9[s-1]>=vv5[s-1] and vv1[s-1]>=vv5[s-1]:
        mcc=[mc,pd.DataFrame(dta9.values[s-1]).T]
        mc=pd.concat(mcc)
        mcc2=[mc2,pd.DataFrame(dta1.values[s-1]).T]
        mc2=pd.concat(mcc2)

mc=mc.reset_index(drop=True)
mc=mc.iloc[10:len(mc),:]
mc=mc.reset_index(drop=True)
mc2=mc2.reset_index(drop=True)
mc2=mc2.iloc[10:len(mc2),:]
mc2=mc2.reset_index(drop=True)
```

```python
mc.loc[1800:(len(mc[2])-1),[0,2]]
```

```python
# 差分近月主力合约和次近主力合约，作图

dttt=td[10:len(td)-1]
difmsm=mc[[6]]-mc2[[6]]
plt.plot(pd.to_datetime(dttt),difmsm,label='diff')
plt.plot(pd.to_datetime(dttt),mc[[6]],label='main')
plt.plot(pd.to_datetime(dttt),mc2[[6]],label='main2')
plt.title('The percentile of Diff_m_m2')
plt.legend(loc = 'upper left')
```

```python
# 对合约序列的差分进行描述性统计
descriptionmsm = pd.DataFrame(index=['value'],columns=['percentile95','percentile80','mean','media', 'percentile20', 'percentile05','standard_deviation'])
descriptionmsm['mean']=np.mean(difmsm.values)
descriptionmsm['media']=np.median(difmsm.values)
descriptionmsm['percentile95']=np.percentile(difmsm.values,95)
descriptionmsm['percentile80']=np.percentile(difmsm.values,80)
descriptionmsm['percentile20']=np.percentile(difmsm.values,20)
descriptionmsm['percentile05']=np.percentile(difmsm.values,5)
descriptionmsm['standard_deviation']=np.std(difmsm.values)
descriptionmsm
```

```python
# 对合约进行ADF检验，检验序列平稳性
from statsmodels.tsa.stattools import adfuller
test = adfuller(pd.Series(difmsm[6]),1)
result = pd.Series(test[0:4], index=['Test Statistic','p-value','Lags Used','Number of Observations Used'])
for key,value in test[4].items():
    result['Confidence level (%s)'%key] = value
pd.DataFrame(result, columns=['diff'])
```

通过检验adf检验结果，发现以1，5，9合约构成的主力和次近主力合约的价差呈现的结果为随机序列，不存在单位根，因而可认为其价差是基于均值的均值回复过程，然而通过观察我们看出，通常豆粕价差的回归会以换月实现，这
