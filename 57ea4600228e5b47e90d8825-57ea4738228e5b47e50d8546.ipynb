# 新功能1：动态universe

自从我们提供了`set_universe`功能后，得到了大家的广泛的使用。运用`set_universe`可以很方便得到一个静态的指数成分股组合，使大家免于写一长串股票代码。然而由于指数成分股通常每半年会调整一次，如果一直用一个静态的指数成分股股票池回测很多年，会给很多策略思想的结果带来一定的扭曲，也就是用到未来信息。


基于此，我们推出`DynamicUniverse`功能,该功能可以实现每天都是一个动态的股票池，并且可以灵活的与静态股票池相结合。

## DynamicUniverse使用方法 

- `DynamicUniverse`和 `set_universe`一样，都是统一用于初始化回测基本关键字`universe`(证券池)。

- `DynamicUniverse`基本用法:

```
universe = DynamicUniverse('HS300')  # 支持动态沪深300指数成分股股票池（返回一个实例，建议只在策略环境中使用）
universe = DynamicUniverse(IndSW.BaiJiuL3)  # 支持动态的行业股票池
universe = DynamicUniverse('HS300', IndSW.BaiJiuL3, '000300.ZICN', 'ZXB')  # 支持指数的不同内部代码（'HS300'和'00300.ZICN'都是沪深300指数）以及多种指数的组合，得到不同指数成分股的动态并集
universe = DynamicUniverse('HS300') + DynamicUniverse('ZXB') + set_universe('SZ50') +  ['000001.XSHE']  # 支持不同指数动态成分股相加，结果是不同指数成分股的动态并集；同样支持动态证券池与静态证券池的相加，得到动态证券池与静态证券池的并集。
```

## 如何用DynamicUniverse写策略

为了配合动态股票池的使用，新增`account.current_universe`概念。

* 对于静态设置`universe`来说(用`set_universe`或直接写一个股票`list`), `account.current_universe`即等于`universe`，`account.universe`为`universe`与你的持仓`account.security_position`的并集，并去除了当天停牌的股票。

* 对于使用了`DynamicUniverse`动态设置`universe`来说，`account.current_universe`为动态的每天`universe`(未去停牌)，`account.universe`为`account.current_universe`与你的持仓`account.security_position`的并集，并去除了当天停牌的股票。

如下图所示：

![](http://storage-uqer.datayes.com/57c4f3f1228e5b6d277c77c3/adc46336-849b-11e6-b565-f8bc124ed898)

其各个部分的意义为:

* 在T时刻，黄色的圆代表此时的持仓,`account.security_position`;蓝色的圆代表此时的动态成份股,`account.current_universe`;绿色的圆代表此时的可交易证券池,`account.universe`。

* `1`部分的含义是，不在持仓中，在动态成份股中，停牌的股票。
* `2`部分的含义是，不在持仓中，在动态成份股中，非停牌的股票。
* `3`部分的含义是，不在持仓中，不在动态成份股中的股票，根据可交易证券池的定义，这部分应当为`空集`。
* `4`部分的含义是，在持仓中，在动态成份股中，停牌的股票。
* `5`部分的含义是，在持仓中，在动态成份股中，非停牌的股票。
* `6`部分的含义是，在持仓中，不在动态成份中，非停牌的股票。
* `7`部分的含义是，在持仓中，不在动态成份股中，停牌的股票。

# 新功能3：refresh rate

鉴于之前`refrensh rate`只支持`int`，满足不了大家个性化的需求，现在`quartz`将支持多种`refresh rate`设置方式。

## refresh rate设置方式


* 日线 
```
    refresh_rate = 1  # 支持int，表示隔refresh_rate个交易日调一次仓  
    refresh_rate = ['2015-01-04', '2015-01-09']  # 支持特定时间点调仓
    
    # 支持每周调仓、每月调仓
    refresh_rate = Weekly(1)  # 每周第1个交易日
    refresh_rate = Monthly(2)  # 每月第2个交易日
    refresh_rate = Weekly(3, -1)  # 每周第3个和倒数第1个交易日
```
    
* 分钟线
```
    refresh_rate = 1  # 支持int，表示每天从09:30开始隔refresh_rate分钟调一次仓
    refresh_rate = ['09:40', '14:50']  # 支持特定分钟点调仓
    refresh_rate = (5, 5)  # 支持tuple，其中第一个参数作用于交易日，第二个参数作用于分钟；表示每5个交易日，在那天从09:30开始每5分钟调一次仓
    refresh_rate = (5, ['09:40', '14:50'])  # 表示每5个交易日，在那天 09:40 和 14:50 时候调仓 
    refresh_rate = (Weekly(1), ['09:40', '14:50'])`  # 表示每周的第1个交易日，09:40 和 14:50 调仓
```  
    
	**总之，分钟线支持日线和分钟线单独设置的任意组合**

# 新功能4：新增组合分析可视化工具

为用户提供更丰富的组合分析的可视化工具，主要包括对于组合（或个股）收益率和持仓的分布分析，包括对于前N大回撤发生的详情统计、收益underwater情况统计、月度收益率分布统计以及策略累计alpha统计。

具体对外提供的函数接口有：

- ``gen_drawdown_table()`` : 生成包含前N大回撤详细信息的DataFrame
- ``plot_drawdown_periods()``：绘出前N大回撤的持续时间分布图
- ``plot_drawdown_underwater()``：绘出在回测周期内underwater分布图
- ``plot_monthly_returns_heatmap()``：绘出回测周期内月度收益热图
- ``plot_monthly_returns_dist()``：绘出回测周期内月度收益分布直方图
- ``plot_pure_alpha()``：绘出日度累积alpha收益图

### 具体用法

- #### 关于输入收益**时间序列**的格式说明
[Pandas Series]
表示组合或个股的当日收益率
Example:
            2015-07-16    -0.012143
            2015-07-17    0.045350
            2015-07-20    0.030957
            2015-07-21    0.004902
Index元素类型为Timestamp或datetime；
若不加特殊说明，PortfolioAnalysis模块中关于收益的时间序列均为该格式
- #### gen_drawdown_table(return, top=10)

	功能：生成前top最大回撤的详细信息，包括回撤百分比、峰值日期、谷底日期、恢复日期以及持续时间

 	输入：return为日收益率的时间序列，top为正整数，表示前top大回撤区间

 	输出：Pandas DataFrame，包含前top大回撤的信息；其列名分别为['net drawdown in %', 'peak date', 'valley date', 'recovery date', 'duration']

- #### plot_drawdown_periods(returns, top=10, ax=None, **kwargs)

	功能：绘出前top最大回撤的区间
    
	输入：return为日收益率的时间序列；top为正整数，表示前top大回撤区间；ax为matplotlib.Axes，表示绘图的子图对象；kwargs表示可以传给plot函数的关键字参数

 	输出：matplotlib.Axes，表示已绘制的子图对象
 
- #### plot_drawdown_underwater(returns, ax=None, **kwargs)

	功能：绘出回撤underwater的情况
    
	输入：return为日收益率的时间序列；ax为matplotlib.Axes，表示绘图的子图对象；kwargs表示可以传给plot函数的关键字参数

 	输出：matplotlib.Axes，表示已绘制的子图对象
- #### plot_monthly_returns_heatmap(returns, ax=None, **kwargs)

	功能：绘制出月度收益率的热图

	输入：return为日收益率的时间序列；ax为matplotlib.Axes，表示绘图的子图对象；kwargs表示可以传给plot函数的关键字参数

 	输出：matplotlib.Axes，表示已绘制的子图对象
- #### plot_monthly_returns_dist(returns, ax=None, **kwargs)
	
    功能：绘制出月度收益率的直方图
    
	输入：return为日收益率的时间序列；ax为matplotlib.Axes，表示绘图的子图对象；kwargs表示可以传给plot函数的关键字参数

 	输出：matplotlib.Axes，表示已绘制的子图对象
- #### plot_pure_alpha(bt, capital_base, ax=None,**kwargs)
	
    功能：绘制出每日完美对冲下的每日超额收益的累计图
    
    输入：bt为quartz.backtest返回值；capital_base为回测的初始资本；ax为matplotlib.Axes，表示绘图的子图对象；kwargs表示可以传给plot函数的关键字参数
    
    输出：matplotlib.Axes，表示已绘制的子图对象

# 变更3：股票筛选器变动对比 quick_backtest调整

现在使用`quick_backtest`可以直接获得`bt`和`perf`，无需调用`perf_parse`。

````python
from quartz import SimulationParameters, TradingStrategy, quick_backtest, get_backtest_data

start = '2015-10-01'
end = '2015-12-01'
benchmark = 'SH50'
universe = set_universe('SH50')
capital_base = 10000
max_history_window = 100

# ---------------回测参数部分结束----------------

# 把回测参数封装到 SimulationParameters 中，供 quick_backtest 使用
sim_params = SimulationParameters(start, end, benchmark, universe, capital_base, max_history_window=max_history_window)
# 获取回测行情数据
data = get_backtest_data(sim_params)
# 运行结果
results = {}
# 调整参数快速回测
for x in range(5):
    def initialize(account):
        account.group = x

    def handle_data(account):
        d = len(account.universe)
        for stock in account.universe[int(d/5.*account.group):int(d/5.*(account.group+1))]:
            order_to(stock, 100)

    # ---------------策略逻辑部分结束----------------

    # 把回测逻辑封装到 TradingStrategy 中，供 quick_backtest 使用
    strategy = TradingStrategy(initialize, handle_data)
    # 回测部分
    bt, perf = quick_backtest(sim_params, strategy, data=data, refresh_rate=refresh_rate)
    # commission = commission,
    # slippage = slippage

    # 保存运行结果
    results[x] = perf['returns']
````

可对比不同参数情况下的收益

````python
import pandas
results = pandas.DataFrame(results)
results.plot()
````
