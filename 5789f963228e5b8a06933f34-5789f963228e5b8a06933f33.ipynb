发现：
1. 加上commission和滑点以后，收益差很多，原来能够达到60%年化，好的情况还剩下45%，如果40%的，剩下不到25%了。所以**`推荐以后回测手续费和滑点必须加上，否则估计的太乐观！`**
2. 模拟交易文档里面写：`没有考虑这两个`，忽略了。所以实盘模拟要自己减少15%，放低要求，这还不算千古跌停时候。。另外大资金的冲击成本这个都没算，所以实际情况要比模拟板差太多，还是别太乐观为好。
3. simple_long_only这个函数只有猜测没有介绍，换了复杂点的**`long_only`**###效果提升很好###，建议大家试试看，我个人猜测里面用了优化和拟合，要不然也不会有回测截至日期这个参数，里面是有预测的。
4. `neutralize()`这个没有详细说明貌似用了行业平衡的权重多因子函数不如不用。用了这个策略效果变差。
5. Fama这个三因子用总市值LCAP比用流通市值LFLO效果怎么测试都好。
6. 幸存者偏差还没加上，加上以后收益效果可能还要差很多，代码参考之前发的[【偏差达4倍】由set_universe("HS300")引起的幸存者偏差对比](https://uqer.datayes.com/community/share/5785eb21228e5b8a0a9327cc)
