**4.2 循环**

```python
a = range(100)
a
```

```python
# for....in.....
#遍历给定集合的所有元素
a = 0
for i in range(101):
    a = a + i
a
```

```python
# while循环，只要满足给定条件，就会不断循环，直到不满足条件
a = 0
i = 1
while i <= 100:
    a = a + i
    i += 1
a
```

```python
# 循环中常伴的两种语法: break, continue。一般与if同时使用
# break: 在循环中，break可以提前结束该循环
for i in range(10):
    if i >= 5:
        break
    print(i)
```

```python
# continue: 在循环中，continue可以跳过当前循环，进行下一次的循环
for i in range(10):
    if i % 2 == 0:
        continue
    print(i)
```

```python
#列表表达式
a = []
for i in range(10):
    if i%2 != 0:
        a.append(i)
a
```

```python
a = [i for i in range(10)]
a
```

**3.2 金融领域常见用法**

```python
#获取20180205沪深300股票当前的行业PE值，并进行排序
df1 = DataAPI.MktEqudGet(tradeDate=u"20180205",secID=u"",ticker=u"",beginDate=u"",endDate=u"",isOpen="",field=u"",pandas="1")
df1 = df1[['secID', 'ticker', 'PE']]

df2 = DataAPI.EquIndustryGet(industryVersionCD=u"010303",industry=u"",secID=u"",ticker=u"",intoDate=u"",field=u"",pandas="1")
df2 = df2[df2['isNew']==1][['secID', 'ticker', 'industryName1', 'industryName2', 'industryName3']]

df3 = DataAPI.IdxConsGet(secID=u"",ticker=u"000300",intoDate=u"",isNew=u"",field=u"",pandas="1")
df3 = df3[df3.isNew==1][['consID', 'consTickerSymbol']]
```

```python
df1['PE']
```

```python
#重新建索引
df1 = df1.set_index('secID', drop=True)
df2 = df2.set_index('secID', drop=True)
```

```python
df3['consID'].tolist()
```

```python
#得到hs300的成分股
hs300 = df3['consID'].tolist()
```

```python
df1.index
```

```python
df1 = df1.ix[hs300]
```

```python
df1
```

```python
len(df1), len(df2)
```

```python
#将两个dataframe合并
df_merge = pd.merge(df1, df2, how='left', on='ticker')
```

```python
df_merge.head()
```

```python
df_merge.groupby('industryName1')
```

```python
#统计行业均值
df_group = df_merge.groupby('industryName1').agg({'PE':'mean'}) 
df_group
```

```python
#排序
df_group = df_group.sort_values('PE', ascending=False)
df_group
```
