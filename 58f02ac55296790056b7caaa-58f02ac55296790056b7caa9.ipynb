```python
from CAL.PyCAL import *
import numpy as np

def get_datelist(startdate,finishdate):   
    data=DataAPI.TradeCalGet(exchangeCD=u"XSHG",beginDate=startdate,endDate=finishdate,field=['calendarDate','isOpen'],pandas="1")
    return data[data['isOpen']==1]['calendarDate'].values.tolist()

period=100
calendar = Calendar('China.SSE')

datelist=get_datelist('2011-01-01','2014-01-01')
for currDate in datelist:
    ptf=set_universe('SH50',currDate)
    endDate = calendar.advanceDate(currDate,'-'+str(1)+'B').toDateTime()
    beginDate = calendar.advanceDate(currDate,'-'+str(period)+'B').toDateTime()  

    tickers = [index[:6] for index in ptf]
    pdData = {index:DataAPI.MktEqudGet(tradeDate=u"",secID=u"",ticker=index,beginDate=beginDate,endDate=endDate,field=[],pandas="1") for index in tickers}

    exsRet = {}
    for index in tickers:
        exsRet[index] = (pdData[index]['closePrice'] - pdData[index]['preClosePrice'])/pdData[index]['preClosePrice']
        exsRet[index].fillna(0)

    tmpList = []
    for index in tickers:
        tmpList.append(exsRet[index].tolist())
    print currDate,'++++++++++++++++++++++++++++++++++++++++++++++'
    print tmpList
    ret = np.mat(tmpList)
    print 'ret.shape=',ret.shape,'ret.values='
    print ret
    print '=============================================='
```
