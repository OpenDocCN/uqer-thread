```python
def calc_short_HL(data):
    data_tmp = data
    """
    对每天的价格进行分类，主要分为4类。
        1、涨线，今日的最高价大于昨日的最高价；今日的最低价大于昨日的最低价 1
        2、跌线，今日的最高价小于昨日的最高价；今日的最低价小于昨日的最低价 2
        3、孕线，今日的最低价大于昨日的最低价；今日的最高价小于昨日的最高价 3
        4、外线，今日的最低价小于昨日的最低价；今日的最高价大于昨日的最高价 4
        设置列名为 category
    """
    data_tmp["prehighestPrice"] = data_tmp["highestPrice"].shift(1)
    data_tmp["prelowestPrice"] = data_tmp["lowestPrice"].shift(1)
    """
    data["tmp_high"] = data["highestPrice"] > data["prehighestPrice"]
    data["tmp_low"] = data["lowestPrice"] < data["prelowestPrice"]
    data["tmp"] = data["tmp_high"].apply(lambda x:int(x)) + data["tmp_low"].apply(lambda x:int(x))
    data["category_4"] = data["tmp"].apply(lambda x : True if x == 2 else False)
    data["category_3"] = data["tmp"].apply(lambda x : True if x == 0 else False)
    """
    category = []
    for i in range(len(data_tmp)):
        tmp_high =  data_tmp.iloc[i]["highestPrice"] > data_tmp.iloc[i]["prehighestPrice"]
        tmp_low =  data_tmp.iloc[i]["lowestPrice"] < data_tmp.iloc[i]["prelowestPrice"]
        if tmp_high and  not tmp_low:
            category.append(1)
        elif not tmp_high and  tmp_low:
            category.append(2)
        elif not tmp_high and  not tmp_low:
            category.append(3)
        elif tmp_high and  tmp_low:
            if data.iloc[i]["closePrice"] >= data.iloc[i]["preClosePrice"]:
                category.append(1)
            else:
                category.append(2)
    data_tmp["category"] = category 
    data_tmp["category_shift1"] = data_tmp["category"].shift(-1)
    date_low_point = []
    date_high_point = []
    date = []
    low_point = []
    high_point = []
    point = []
    triger = -1
    for i in range(len(data_tmp)-1):

        j = i+1
        today = data_tmp.iloc[j]["category"]
        tomorrow = data_tmp.iloc[j]["category_shift1"]
        yesterday = data_tmp.iloc[i]["category"]

        if today == 1:
            if tomorrow == 1:
                pass
            elif tomorrow == 2:
                tmp = data_tmp.iloc[j]["highestPrice"]
                date_high_point.append(j)
                high_point.append(tmp)
                date.append(j)
                point.append(tmp)
            elif tomorrow == 3:
                triger = -1
        elif today == 2:
            if tomorrow == 1:
                tmp = data_tmp.iloc[j]["lowestPrice"]
                date_low_point.append(j)
                low_point.append(tmp)
                date.append(j)
                point.append(tmp)
            elif tomorrow == 2:
                pass
            elif tomorrow == 3:
                triger = -1
        elif today == 3:
            if tomorrow == 1:
                if yesterday == 1:
                    pass
                elif yesterday == 2:
                    tmp = data_tmp.iloc[i]["lowestPrice"]
                    date_low_point.append(i)
                    low_point.append(tmp)
                    date.append(i)
                    point.append(tmp)
                    triger = 999
                elif yesterday == 3:
                    # 当情况是3的时候看之前的状态是什么
                    triger_tmp = j+triger
                    category_pre = data_tmp.iloc[triger_tmp]["category"]
                    if category_pre == 1:
                        pass
                    elif category_pre == 2:
                        tmp = data_tmp.iloc[triger_tmp]["lowestPrice"]
                        date_low_point.append(triger_tmp)
                        low_point.append(tmp)
                        date.append(triger_tmp)
                        point.append(tmp)
                        triger = 999

            elif tomorrow == 2:
                if yesterday == 1:
                    tmp = data_tmp.iloc[i]["highestPrice"]
                    date_high_point.append(i)
                    high_point.append(tmp)
                    date.append(i)
                    point.append(tmp)
                    triger = 999
                elif yesterday == 2:
                    pass
                elif yesterday == 3:
                    # 当情况是3的时候看之前的状态是什么
                    triger_tmp = j+triger
                    category_pre = data_tmp.iloc[triger_tmp]["category"]
                    if category_pre == 1:
                        tmp = data_tmp.iloc[triger_tmp]["highestPrice"]
                        date_high_point.append(triger_tmp)
                        high_point.append(tmp)
                        date.append(triger_tmp)
                        point.append(tmp)
                        triger = 999
                    elif category_pre == 2:
                        pass
            elif tomorrow == 3:
                triger -= 1
    
    
    """
    取三个短期高低点，因为肯定是两高一低，或者两低一高。用第三个去和前面两个比较，所以有这么几种情况
    1、如果第三个是高点的话，如果它高于第一个高点，则第二个低点为中期低点！
    2、如果第三个是低点的话，如果它低于第一个低点，则第二个高点为中期高点！
    3、如果第三个是高点的话，如果它低于第一个高点的话，说明中期跌势继续延续，记录下第二个低点的情况，因为它很可能成为中期低点，当它被跌破的时候警报解除
    4、 
    """
    high_point_zip = zip(date_high_point, high_point)
    low_point_zip = zip(date_low_point, low_point)
    point_zip = zip(date, point)
    middle_high = []
    middle_low = []
    middle_high_date = []
    middle_low_date = []
    middle_date = []
    middle_date_point = []
    category = 0             # category=1为涨势； category=2为跌势
    tmp_type = 0       # type=1为高点； type=2为低点
    tmp_point = 0
    tmp_dare = 0
    tmp_category = 0
    triger = False
    for i in range(len(point_zip)-2):

        if point_zip[i] in high_point_zip:
            tmp_type = 1
        else:
            tmp_type = 2
        day_1 = point_zip[i][1]
        day_2 = point_zip[i+1][1]
        day_3 = point_zip[i+2][1]
        if tmp_type == 1:
            # 判断是否为中期低点
            if day_3 >  day_1:
                # 高点突破，判断之前是否为涨势
                if category <> 1:
                    # 之前不是涨势，确定涨势，同时记录中期低点
                    middle_low.append(day_2)
                    middle_low_date.append(point_zip[i+1][0])
                    middle_date_point.append(day_2)
                    middle_date.append(point_zip[i+1][0])
                    category = 1
                elif category == 1:
                    # 之前为涨势，pass
                    pass
            elif day_3 <=  day_1:
                # 高点未突破，记录前期的低点，可能为中期低点
                pass
        elif tmp_type == 2:
            if day_3 <  day_1:
                # 高点突破，判断之前是否为涨势
                if category <> 2:
                    # 之前不是涨势，确定涨势，同时记录中期低点
                    middle_high.append(day_2)
                    middle_high_date.append(point_zip[i+1][0])
                    middle_date_point.append(day_2)
                    middle_date.append(point_zip[i+1][0])
                    category = 2
                elif category == 2:
                    # 之前为涨势，pass
                    pass
            elif day_3 >=  day_1:
                # 高点未突破，记录前期的低点，可能为中期低点
                pass
    short_date = date
    short_point = point
    return short_date, short_point, middle_date,middle_date_point
```

```python
def plot_HL(ticker, beginDate, endDate=datetime.date.today()):
    quotes = DataAPI.MktEqudAdjGet(ticker=ticker, beginDate=beginDate, endDate=endDate, isOpen='1',field=["secShortName","ticker","openPrice","closePrice","highestPrice","lowestPrice","tradeDate","dealAmount","isOpen","preClosePrice"])
    short_date, short_point, middle_date,middle_date_point = calc_short_HL(quotes)
    final_point_x = [short_date[-1], len(quotes)-1]
    final_point_y = [short_point[-1], quotes.iloc[-1]["closePrice"]]
    
    
    fig = plt.figure(figsize=(12.5,5))
    #fig.set_tight_layout(True)
    ax1 = fig.add_axes([1, 1, 1, 1])
    ax1.set_title(quotes.ix[0, 'tradeDate'])
    ax1.set_axisbelow(True)
    ax1.plot(short_date,short_point,"b")
    ax1.plot(middle_date,middle_date_point,"r")
    ax1.plot(final_point_x,final_point_y,"y")
    ax1.grid(True)
    ax1.set_xlim(-1, len(quotes)+1)
    ax1.set_title(quotes['secShortName'][1].decode('utf-8'), fontproperties=font, fontsize=15, loc='left', color='r')

    ax1.set_xticks(range(0,len(quotes), 15))
    s1 = ax1.set_xticklabels([quotes.ix[index, 'tradeDate'] for index in ax1.get_xticks()])
    ax1.annotate('short_HL', xy=(len(quotes)-32, ax1.get_yticks()[-1]), color='b', fontsize=10)
    ax1.annotate('Middle_HL', xy=(len(quotes)-15, ax1.get_yticks()[-1]), color='r', fontsize=10)

    return fig
```
