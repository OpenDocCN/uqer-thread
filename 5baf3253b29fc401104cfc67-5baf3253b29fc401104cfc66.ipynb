```python
## 获得上述指数的日行情。
beginDate='20060101'
endDate='20180402'
indexData=DataAPI.MktIdxdGet(beginDate=beginDate,endDate=endDate,indexID=u"",ticker=IndexCode,field=u"",pandas="1")
## 求得每个交易日的收益率。从而求得每个月的波动率。
indexData['tradeMonth']=indexData.tradeDate.apply(lambda x:x[:7])
indexData.head()
```

```python
indexData['intraDayRtn']=(indexData.closeIndex-indexData.openIndex)/indexData.openIndex
indexData['lastNightRtn']=(indexData.openIndex-indexData.preCloseIndex)/indexData.preCloseIndex
indexData.head()
```

```python
CalRtn=indexData.groupby(['tradeMonth','indexID']).apply(lambda x:x[['intraDayRtn','lastNightRtn']].iloc[-15:].sum())

def rank_return(x):
    ## lastNightRtn越大，排名值越大
    ## lastNightRtn要选【0-6】的
    x['lastNightRank']=x['lastNightRtn'].rank()
    ## intraDayRtn值越小，排名值越大
    ## intraDayRtn要选【0-6】的
    x['intraDayRank']=x['intraDayRtn'].rank(ascending=False)
    return x
    pass

def rank_finalScore(x):
    ## score值越小，排名值越小
    ## score要选【0-6】的
    x['finalRank']=x['score'].rank(ascending=True)
    return x
    pass


rankRtn=CalRtn.groupby(level=[0]).apply(rank_return)
rankRtn['score']=rankRtn.lastNightRank+rankRtn.intraDayRank+rankRtn.intraDayRtn
finalRank=rankRtn.groupby(level=[0]).apply(rank_finalScore)
finalRank
```
