# 期货蛛网模型

```python
import numpy as np
import pandas as pd
import scipy.stats as st
import scipy.optimize as opm
from sklearn import datasets, linear_model
import statsmodels.api as sm

import datetime
import time
from CAL.PyCAL import *
cal = Calendar('China.SSE')
```

```python
future_ticker=['J1801',]
bdate="20170501"
edate="20170918"
ndate="2017-09-15"
```

```python
Ldata=DataAPI.MktFutMLRGet(secID=u"",ticker=future_ticker,beginDate=bdate,endDate=edate,field=u"",pandas="1")
Sdata=DataAPI.MktFutMSRGet(secID=u"",ticker=future_ticker,beginDate=bdate,endDate=edate,field=u"",pandas="1")
Tdata=DataAPI.MktFutMTRGet(secID=u"",ticker=future_ticker,beginDate=bdate,endDate=edate,field=u"",pandas="1")
L=Ldata[Ldata['tradeDate']==ndate][:]
S=Sdata[Sdata['tradeDate']==ndate][:]
T=Tdata[Tdata['tradeDate']==ndate][:]
T
```

```python
L1=L[['partyShortName','longVol','CHG']]
L1['longCHG']=L1['CHG']
L1=L1.drop('CHG',axis=1)
S1=S[['partyShortName','shortVol','CHG']]
S1['shortCHG']=S1['CHG']
S1=S1.drop('CHG',axis=1)
T1=T[['partyShortName','turnoverVol','CHG']]
T1['turnoverCHG']=T1['CHG']
T1=T1.drop('CHG',axis=1)
```

```python
dall=pd.DataFrame()
dall=pd.merge(L1,S1,how='outer',on='partyShortName')
dall=dall.merge(T1,how='outer',on='partyShortName')
dall['NaNs']=dall.isnull().sum(axis=1)
dall0=dall[dall['NaNs']==0]
dall1=dall[dall['NaNs']!=0]
new0=pd.DataFrame()
if len(dall1)>0:
    new0=dall1.iloc[:,1:8].apply(lambda x: x.sum())
    new0=new0.to_frame().T
    new0.insert(0,'partyShortName','othersNaN')
dall2=dall0.append(new0)
dall2.dropna(inplace=True)
dall2['turnover%']=dall2['turnoverVol']/dall2['turnoverVol'].sum()
dall2big=dall2[dall2['turnover%']>0.01].copy()
dall2small=dall2[dall2['turnover%']<=0.01].copy()
new1=pd.DataFrame()
if len(dall2small)>0:
    new1=dall2small.iloc[:,1:9].apply(lambda x: x.sum())
    new1=new1.to_frame().T
    new1.insert(0,'partyShortName','turnoversmall')
dall3=dall2big.append(new1)
dall3
```

```python
dall3['Stat']=(dall3['longVol']+dall3['shortVol'])/dall3['turnoverVol']
dall3['Ts_CHG']=(dall3['longCHG']-dall3['shortCHG'])/(abs(dall3['longCHG'])+abs(dall3['shortCHG']))
dall3['Ts_Vol']=(dall3['longVol']-dall3['shortVol'])/(dall3['longVol']+dall3['shortVol'])
dall3['L-S']=dall3['longVol']-dall3['shortVol']
dall3['dL-dS']=dall3['longCHG']-dall3['shortCHG']
dall3.sort('Stat',ascending=False,inplace=True)
dall3
```

```python
Statall=(dall3['longVol'].sum()+dall3['shortVol'].sum())/dall3['turnoverVol'].sum()

BVolITs=dall3[dall3['Stat']>Statall]['longVol'].sum()
SVolITs=dall3[dall3['Stat']>Statall]['shortVol'].sum()
BVolUTs=dall3[dall3['Stat']<=Statall]['longVol'].sum()
SVolUTs=dall3[dall3['Stat']<=Statall]['shortVol'].sum()
VolITS=(BVolITs-SVolITs)/(BVolITs+SVolITs)
VolUTS=(BVolUTs-SVolUTs)/(BVolUTs+SVolUTs)

BCHGITs=dall3[dall3['Stat']>Statall]['longCHG'].sum()
SCHGITs=dall3[dall3['Stat']>Statall]['shortCHG'].sum()
BCHGUTs=dall3[dall3['Stat']<=Statall]['longCHG'].sum()
SCHGUTs=dall3[dall3['Stat']<=Statall]['shortCHG'].sum()
CHGITS=(BCHGITs-SCHGITs)/(abs(BCHGITs)+abs(SCHGITs))
CHGUTS=(BCHGUTs-SCHGUTs)/(abs(BCHGUTs)+abs(SCHGUTs))

ans=[VolITS,VolUTS,CHGITS,CHGUTS]
print(ans)
```
