然后我们要看具体行业内的具体公司。核心函数DataAPI.EquIndustryGet

要点，该函数返回的是历史上的所有数据，可以通过下面的示例看出来

```python
# 尝试一：阵亡的案例。原因，退市公司的行业分类它永远没有出去的一天。
# 之所以写出来，是因为API文档这种东西永远无法事无巨细的写全，操作过才知道。
ind = DataAPI.EquIndustryGet(industryVersionCD='010303')
ind[ind['outDate'].isnull()].head(10)
# 注，000003.XSHE 早已经退市了
```

```python
# 尝试二：换种思路。因为上述函数中有个intoDate关键字，输入日期，可以获取这一天股票的行业信息，输入格式“YYYYMMDD”,可空。结局又阵亡。所以给大家排除了两种错误的使用方法。
import datetime
current = datetime.datetime.now().strftime('%Y%m%d')
ind = DataAPI.EquIndustryGet(industryVersionCD='010303',intoDate=current)
ind.head(10)
# 000003.XSHE 你怎么还在
```

```python
# 于是最后一种方法是暴力筛选

#先获取最终的一类行业划分，但是包括过去的股票
ind = DataAPI.EquIndustryGet(industryVersionCD='010303')
ind = ind[ind['outDate'].isnull()]

equl1 = DataAPI.EquGet(equTypeCD=u"A",secID=u"",ticker=u"",listStatusCD=u"L",field=u"",pandas="1")
equl2 = DataAPI.EquGet(equTypeCD=u"A",secID=u"",ticker=u"",listStatusCD=u"S",field=u"",pandas="1")
equlist = equl1.append(equl2,ignore_index=True)

current_ind = ind[ind['secID'].isin(equlist['secID'])].copy()
current_ind.head(10)
```

```python
# 上面已经有点点点了，表示列数太多显示不起来了，我们排除部分不需要的数据。从代码的角度讲，是不是该鼓励用这种，而不是用del直接删？有人解答下吗？
current_ind.drop(['ticker','outDate','exchangeCD'],axis=1,inplace=True)
```

```python
current_ind[['secID','industryName1','industryID1']].groupby(['industryID1','industryName1']).count().sort('secID')
# 数目上是不是有点不合理呢？前面说了，不合理需要加上市值看啊。有些行业叫国情啊。
```

```python
# 随便指定一天，获取市值。这里又是一个数据上需要说明的问题，万科明明是被暂停上市了，但是当天还是有市值的，通联的处理做法是将暂停上市的数据OHLC设置为0,0,0,0.所以我们就不用手动管这个暂停上市的问题了。
marketValue = DataAPI.MktEqudGet(tradeDate=u"20160624",secID=u"",ticker=u"",beginDate=u"",endDate=u"",isOpen="",field=u"secID,marketValue",pandas="1")
marketValue.head(10)
```

```python
import pandas as pd
df = pd.merge(current_ind,marketValue,on='secID')
df.head(10)
```

```python
import numpy as np
```

```python
df[['secID','marketValue','industryName1','industryID1']].groupby(['industryID1','industryName1']).agg({'secID':'count','marketValue':np.sum}).sort('marketValue')
# 虽然在数目这个维度，不尽合理；但是加上市值的话是比较合理的。行业市值加和最高不超过20倍。
```

```python
# 看一波二级行业分类了。这个时候行业因子在使用中要小心了，已经有行业数据为一个手都能数得过来的了。这个时候一个有用的借鉴是，将上面的市值加和改成算术平均。
# 这样你就知道市值上的傻大个主要分布在哪些行业中了。由于市值（或可称为‘规模’）也是个重要的风险因子，二级行业分类对于寻找傻大个比较有用。
# 比如我们发现在银行、石油、证券、保险是典型的傻大个。但是房地产却不是，这是因为有些上市的地区性房地产公司规模不是很大，拖累了整体。
# 在我们可以接受的认知范围，第一类（太少）和第二类（太多）划分，通常可以给我们一个初步预判，整体行业行不行，要不要在这个领域的企业中投资。
df[['secID','marketValue','industryName2','industryID2']].groupby(['industryID2','industryName2']).agg({'secID':'count','marketValue':np.mean}).sort('marketValue')
```

```python
# 注意：三级行业分类已经多达225项，已经完全超出了我们能接受的范围。在这样的细分领域，我们要担心多种了。
# 第一，分类变更不及时，企业已经发生行业变更，但是数据还没更新。
# 如果不存在上面的情况，三级分类可用于相似企业的基本面对比。这才是三级行业使用的真谛，因为要对比的前提是有对比性，而如果越细分，通常可以确定企业所在的领域越相近。除非申万傻了，分类错了。或者说，某些名义上近似的行业类别实际上具有不可替代性，典型是产品的客户群体是高端还是低端。但这个借鉴意义是极好的。
df[['secID','marketValue','industryName3','industryID3']].groupby(['industryID3','industryName3']).agg({'secID':'count','marketValue':np.sum}).sort('marketValue')
```

# 总结

一级分类，和二级分类最符合上面的三点原则。通常来讲，一级太少，二级太多，三级已经超过了人类可以接受的范围。

一级分类和二级分类可以用来做行业大筛选，比如，我觉得这个行业不行了，那么我干嘛还要投资里面的股票。

三级分类的好处在于，如果组里面公司数量比较多的话，第一，它们是良好的基本面对比组（缺陷见前述）；第二，这个行业风险也大（红海竞争）。
