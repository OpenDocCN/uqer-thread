## 突发奇想
### 凡有效的动量，均为箱体突破
昨天突然想到，其实凡是有效的动量，都应该是箱体突破的
动量范围大于箱体突破，可以捕捉更多信号，但是如果**连前期高点都未突破极有可能是无效突破**（比如猴市反复冲高回落）
所以感觉上来讲，箱体突破比动量更加有效
简单起见，先用来做HS300择时试试

**简单箱体**
1，N日内的最高价，最低价成为N日箱顶底部
2，平行箱体前周期越长，后续涨幅越大（暂未考虑）
3，简单起见，暂不考虑斜箱体（暂未考虑）

## 第一版：简单箱体
买入条件：昨日收盘价格大于20日最高收盘价
卖出条件：相对20日最高点回撤3%
每日刷新

** 后续优化：**
1，长短箱体搭配
2，短期休眠池（应对冲高回落时，后续再次突破，当前机制捕捉太慢）
3，长期底徘徊，上升+突破（可让盘整时间控制仓位，盘整时间长的提高仓位）
4，资金平滑处理

### 效果并不理想

3，后面震荡市表现一般。但是可以理解的，反复震荡造成假突破，可能导致买入后触发止损卖出带来亏损。

## 第二版：长短箱体搭配
保守起见，长箱用来建仓，短箱用来止损

### 结果并不理想
尤其16-06-〉16-09这段时间的小上波并未抓住

之所以有这个问题，是由于我们对**箱体定义有问题**
简单箱体：n日最高价，n日最低价，这会导致**上一波已经结束的波段的极值，对此波带来影响。**
**比如15年9月的那一小波上涨，看起来很明显但实际并未抓住，就是由于15年8月虽大跌，但是导致后续买入门槛变高**
**合理的切分箱体，应该是按照波峰波谷切分（max，min）**

## 第三版：重新定义箱体
1，转折点（拐点）:今天价格为前N日最高价/最低价，保存此信息（箱子头底）。介于高低之间，不处理，
此时有个问题，如果这N天是连续突破（比如一直涨停），那么极值点就会连起来，一天一个，这也不符合期望。
期望的拐点时价格有转折的极值点。为了解决此问题，增加定义
2，转折点覆盖：如果明后天（N天内），又有新高创立，则删除今日的转折点，变为普通点

### 消除未来后效果预览

```python
#信号右移，等于手工延迟信号，消除未来函数
import numpy as np
import pandas as pd
#todo股票需修改为adj前复权

'''stk_data=DataAPI.MktEqudAdjGet(tradeDate='',secID=u"159919.XSHE",ticker=u"",isOpen="",beginDate=u"20150101",endDate=u"20170101",field=u"",pandas="1")'''

stk_data=DataAPI.MktFunddGet(tradeDate=u"",secID=u"159919.XSHE",ticker=u"",beginDate=u"20150101",endDate=u"20170101",field=u"",pandas="1")

stk_data['MA5']=pd.rolling_mean(stk_data['closePrice'],window=5)
#提取极值点
max_price=[0 for i in range(0,len(stk_data))]
min_price=[0 for i in range(0,len(stk_data))]
extre_range=5
for i in xrange(extre_range,len(stk_data)):
    if stk_data['closePrice'][i] == stk_data['closePrice'][i-extre_range:i+1].max():
        max_price[i]=1
        max_price[i-extre_range:i]=[0 for ii in range(0,extre_range)]
    elif  stk_data['closePrice'][i] == stk_data['closePrice'][i-extre_range:i+1].min():
        min_price[i]=1
        min_price[i-extre_range:i]=[0 for ii in range(0,extre_range)]
#箱头底和并为一个序列
extre_price=map(lambda (a,b):a+b,zip(min_price,max_price))
#右移信号消除未来函数影响
stk_data['extre_price']=(stk_data['closePrice']*extre_price).shift(extre_range)
stk_data['extre_price']=(stk_data['extre_price'].apply(lambda x:np.nan if x<0.01 else x)).fillna(method='pad')
stk_data[['closePrice','extre_price']].plot(figsize=(12,6))
```

### 生成对应交易序列
1，对每个点（每日），最近的折线的上下沿就是自己的所属箱体
2，买入条件为突破上箱体，（todo，同时也可对折线长度要求下，避免猴市频繁操作）
3，卖出条件为x日高点回撤达到0.97，或跌破下箱体

交易序列最合理的是在回测时生成，但考虑到回测时如果是股票，无法取得前复权价格（为了避免隐藏未来函数，所以回测时个人一般不用api取价格信息)，那样就不好比对箱体位置。
理论上目前已经没有未来信息了，偷个懒，直接生成交易信号了。
矿友有兴趣可以自行验证下，修改为回测时动态生成交易序列。

### 交易序列预览图如下
最高横线是持有的时间，0和最下的都没持有

```python
#生成买卖信号
stk_data['buy_sell']=[0 for li in range(0,len(stk_data))]
high=0 #stk_data['extre_price'][fi-extre_range:fi].max()
low=0 #stk_data['extre_price'][fi-extre_range:fi].min()
for fi in range(extre_range,len(stk_data)):
    if np.isnan(stk_data['extre_price'][fi]) or stk_data['extre_price'][fi]<0.01 :
        continue
    #昨今不等说明有竖线，需要修改箱体上下沿
    if stk_data['extre_price'][fi-1] != stk_data['extre_price'][fi]:
        high=stk_data['extre_price'][fi-1:fi].max()
        low=stk_data['extre_price'][fi-1:fi].min()
    #买入条件为高于high1.03
    if stk_data['closePrice'][fi]>high*1.03:
        stk_data['buy_sell'][fi]=1
    elif stk_data['closePrice'][fi]<low or stk_data['closePrice'][fi]/high<0.97:
        stk_data['buy_sell'][fi]=-1
#画图好看些
stk_data['buy_sell_show']=stk_data['buy_sell']*stk_data['closePrice'].mean()
stk_data[['buy_sell_show','extre_price','closePrice','MA5']].plot(figsize=(12,6))    
```
