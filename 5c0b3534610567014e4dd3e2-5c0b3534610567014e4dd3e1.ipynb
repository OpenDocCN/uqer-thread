```python
import matplotlib.pyplot as plt
import numpy as np 
import pandas as pd
import statsmodels.api as sm

stockdata=DataAPI.MktAHCompdGet(beginDate=u'20060101',endDate=u'20161231')
stock_ticker_list=stockdata['tickerA']
stock_ticker_list=np.unique(stock_ticker_list)
stockdata=DataAPI.MktAHCompdGet(ticker=stock_ticker_list)
stockdata=stockdata.dropna(axis=0,how='any')
print(len(stock_ticker_list))

coefficient=[]
const=[]
for i in range(0,len(stock_ticker_list)-1):
    ticker_temp=stock_ticker_list[i]
    stockdata_temp=stockdata[stockdata['tickerA']==ticker_temp]
    X=sm.add_constant(stockdata_temp['closePriceA'])
    X.columns=['const','coefficient']
    Y=stockdata_temp['closePriceHA']
    reg_res=sm.OLS(Y,X).fit()
    const.append(reg_res.params.const)
    coefficient.append(reg_res.params.coefficient)
plt.hist(coefficient,bins=15,normed=True)
plt.show()
plt.hist(const,bins=15,normed=True)
plt.show()

ticker_temp='600585'
stockdata_temp=stockdata[stockdata['tickerA']==ticker_temp]
tradingday=len(stockdata_temp['tradeDate'])
x=sm.add_constant(stockdata_temp['closePriceA'])
x.columns=['const','coefficient']
y=stockdata_temp['closePriceHA']
const_temp=[]
coefficient_temp=[]
for N in range(180,tradingday-1):
    X=x[N-180:N]
    Y=y[N-180:N]
    reg_res=sm.OLS(Y,X).fit()
    const_temp.append(reg_res.params.const)
    coefficient_temp.append(reg_res.params.coefficient)
plt.plot(const_temp)
plt.show()
plt.plot(coefficient_temp)
plt.show()
```
