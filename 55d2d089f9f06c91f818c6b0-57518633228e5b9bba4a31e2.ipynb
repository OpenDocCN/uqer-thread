1.2. 过度优化的结果
------------------------
> 我们来想想看看，如何”优化“我们的结果？

首先我们先分析一下历史上的价格趋势：

```python
data = DataAPI.MktEqudAdjGet(ticker='600000', beginDate='20110101', endDate='20140101', field=['closePrice'])
data.plot(figsize=(16,8))
```

```python
print("历史最高价：%.4f" % data.max()[0])
print("历史最低价：%.4f" % data.min()[0])
```

这个时候我们想到了一个最简单的注意，如果我们可以买在一个一个价格的波谷，然后卖在随后的价格波峰上，就可以获得风险收益比最高的策略了吗？

同样的，我们可以看一下交易次数，这次只有两次！我们成功的买在了波谷，卖在了波峰！！

```python
bt[bt.blotter != [[]]*len(bt)]
```

1.3. 总结
-----------------------

这里显示的问题实际上是多目标优化时候经常面临的问题。

> 经验准则1：如果你希望你的最终优化结果是稳定的且具有解释力的，那么你的优化目标的个数，应该要显著高于你的自由变量的个数。

在上面的展示的问题中，最后的交易次数只有两次。这意味着对于优化器来说，它需要优化的目标只是一次买入和一次卖出，而同时自由变量两个（``buySignal``以及``sellSignal``)。这样的最优化问题，几乎总是存在完美解。但是显然的，这样的结果很可能是不稳定的，而且一般来说不可能有解释力。

> 经验准则2：优化时的参数应该显著低于回测期内的交易次数。

实际上，我们在第二步的优化时，挑选波峰和波谷的时候，已经在使用未来数据。但是如果这一挑选过程是用优化器完成的时候，这个历史数据就使用的很隐蔽。

> 经验准则3：当使用优化器进行策略参数优化的时候，策略编写者已经在使用未来数据了。
