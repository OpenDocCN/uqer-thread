# fund_analysis-new

### 成功的主动管理依赖于高质量信息
---
- 高质量信息的特征有：`高信息率（IR），高信息系数（IC）`以及 `足够高的广度（Breath）`，公式为：
$$IR =IC * \sqrt{Breath}$$

- 但是在实际中，即使是有能力的基金经理也不会每月，每年都超过基准。我们无法避免历史业绩数据中的噪声；并且时间尺度越短，噪声的影响就越大。


### 如果业绩真的有持续性，什么可以预测业绩？
---
`猜测`
- 在高信息率的前提下，从 `持有更多的股票` 或者 `更高的换手率` 的基金经理那里看到业绩的持续性

```python
# 获取基金的daily_return
def fund_daily_return(fundTicker, benchmarkTicker, begin, end):

    daily_ret = pd.DataFrame()   
    data = DataAPI.FundNavGet(ticker=fundTicker,beginDate=begin,endDate=end,
                              field=u"secShortName,endDate,ADJUST_NAV",pandas="1")
    daily_ret['endDate'] = data['endDate'][1:]
    daily_ret[data.iat[0,0]] = data['ADJUST_NAV'].values[1:]*1.0/data['ADJUST_NAV'].values[:len(data)-1]-1 
    
    daily_ret_2 = pd.DataFrame()
    data_2 = DataAPI.MktIdxdGet(ticker=benchmarkTicker,beginDate=begin,endDate=end,
                                field=u"secShortName,tradeDate,closeIndex",pandas="1")
    daily_ret_2['endDate'] = data_2['tradeDate'][1:]
    daily_ret_2[data_2.iat[0,0]] = data_2['closeIndex'].values[1:]*1.0/data_2['closeIndex'].values[:len(data_2)-1]-1
    
    daily_ret = pd.merge(daily_ret, daily_ret_2, how='inner', on='endDate')
    daily_ret.set_index('endDate',inplace=True) 
    return daily_ret


# 获取基金的最大回撤
def cal_maxdrawdown(data): 
    
    if isinstance(data, list):
        data = np.array(data)
    if isinstance(data, pd.Series):
        data = data.values    
    def get_mdd(values): 
        dd = [values[i:].min() / values[i] - 1 for i in range(len(values))]
        return abs(min(dd))
    if not isinstance(data, pd.DataFrame):
        return get_mdd(data)
    else:
        return data.apply(get_mdd)

    
# 获取基金的评价指标   
def cal_indicators(df_daily_return):
    
    freq_ = 250
    df_cum_value = (df_daily_return + 1).cumprod()
    res = pd.DataFrame(index=['年化收益率','年化标准差','贝塔','夏普比率','信息比率','最大回撤'], 
                       columns=df_daily_return.columns, data=0.0)
    
    res.loc['年化收益率'] = (df_daily_return.mean() * freq_).apply(lambda x: '%.2f%%' % (x*100))
    res.loc['年化标准差'] = (df_daily_return.std() * np.sqrt(freq_)).apply(lambda x: '%.2f%%' % (x*100))
    res.loc['贝塔', daily_ret.icol(1).name] = 1
    res.loc['贝塔', daily_ret.icol(0).name] = np.round((np.cov(df_daily_return[daily_ret.icol(0).name],
                                             df_daily_return[daily_ret.icol(1).name])[0,1] 
                                             / np.cov(df_daily_return[daily_ret.icol(1).name])), 2)
    
    res.loc['夏普比率'] = (df_daily_return.mean() / df_daily_return.std() * np.sqrt(freq_)).apply(
                                                                        lambda x: np.round(x, 2))
    
    res.loc['信息比率', daily_ret.icol(0).name] = np.round(((df_daily_return[daily_ret.icol(0).name].mean()
                                                         - df_daily_return[daily_ret.icol(1).name].mean())  
                                                         / (df_daily_return[daily_ret.icol(0).name]-
                                        df_daily_return[daily_ret.icol(1).name]).std() * np.sqrt(freq_)), 2)
    
    res.loc['最大回撤'] = cal_maxdrawdown(df_cum_value).apply(lambda x: '%.2f%%' % (x*100))
    return res


# 股票格式转换
def sec2ticker(sec):
    
    universe = DataAPI.EquGet(equTypeCD=u"A",listStatusCD="L,S,DE,UN",field=u"ticker,secID",pandas="1") 
    universe = universe[['secID','ticker']]
    universe = dict(universe.set_index('secID')['ticker'])
    if isinstance(sec, list):
        res = []
        for i in sec:
            if i in universe:
                res.append(universe[i])
            else:
                print i 
        return res
    elif isinstance(sec, dict):
        res = {}
        for i in sec:
            if i in universe:
                res[universe[i]] = sec[i]
            else:
                print i
        return res
    elif isinstance(sec, pd.Series):
        res = {}
        for i in sec.index:
            if i in universe:
                res[universe[i]] = sec[i]
            else:
                print i
        return pd.Series(res)
    else:
        raise ValueError('sec should be list or dict or series！')


# 统计不同类型的基金收益
def fund_type_ret(start_="20171231", end_="20180706", report_date="20180331"):
    
    # 所有基金数据
    fund_sec = DataAPI.SecIDGet(assetClass=u"F",pandas="1")
    fund_info = DataAPI.FundGet(secID=fund_sec.secID,pandas="1")
    # 去除分级基金
    fund_leverage = DataAPI.FundLeverageInfoGet(secID=u"",ticker=u"",exchangeCDLeverage=u"XSHG,XSHE",field=u"",pandas="1")
    fund_info = fund_info[[x not in fund_leverage.secIDLeverage.values for x in fund_info.secID.values]]
    # E-股票型，H-混合型
    fund_info = fund_info[[x in ['H','E'] for x in fund_info.category]]
    # O-开放式，C-封闭式
    fund_info = fund_info[[x not in ['C'] for x in fund_info.operationMode]]
    # isFof-是否为FOF基金  
    # isQdii-是否为QDII基金
    # isGuarFund-是否为保本基金
    fund_info = fund_info[(fund_info.isFof==0) & (fund_info.isQdii==0) & (fund_info.isGuarFund==0)]
    # 去除海外基金
    fund_info = fund_info[['标普' not in x and '纳斯达克' not in x and '美国' not in x for x in fund_info.secShortName]]
    fund_info = fund_info[['标普' not in str(x) and '纳斯达克' not in str(x) and '美国' not in str(x) for x in fund_info.idxShortName]]
    fund_info = fund_info[['标普' not in str(x) and '纳斯达克' not in str(x) and '美国' not in str(x) for x in fund_info.perfBenchmark]]
    fund_info = fund_info[['MSCI' not in str(x) and '标准普尔' not in str(x) and '全球' not in str(x) for x in fund_info.perfBenchmark]]
    fund_info = fund_info[['欧洲' not in str(x) and '摩根斯坦利' not in str(x) and '富时' not in str(x) for x in fund_info.perfBenchmark]]
    fund_info = fund_info[['德国' not in str(x) and '恒生' not in str(x) and '香港' not in str(x) for x in fund_info.perfBenchmark]]
    fund_info = fund_info[['海外' not in str(x) and '恒生' not in str(x) and '香港' not in str(x) for x in fund_info.perfBenchmark]]
    # ----------------------------- 量化基金 -----------------------
    # 指数增强
    fund_quant_EI_info = fund_info[[('增强' in x or '指数增强' in x) for x in fund_info.secShortName]]
    # 绝对收益
    fund_quant_hedge_info = fund_info[[('对冲' in x or '绝对收益' in x) for x in fund_info.secShortName]]  
    # 量化基金
    fund_quant_info = fund_info[[('量化' in x or '多因子' in x or 
                                  '对冲' in x or '绝对收益' in x or 
                                  '阿尔法' in x or '指数增强' in x) for x in fund_info.secShortName]]
    fund_quant_info = fund_quant_info.append(fund_info[fund_info.secID=='229002.OFCN'])
    fund_quant_info = fund_quant_info.append(fund_info[fund_info.secID=='202019.OFCN'])
    fund_quant_info = fund_quant_info.append(fund_info[fund_info.secID=='001637.OFCN'])
    fund_quant_info = fund_quant_info.append(fund_quant_EI_info).append(fund_quant_hedge_info).drop_duplicates('secID')
    # 量化主动
    fund_quant_active_info = fund_quant_info[[x not in fund_quant_hedge_info.secID.values and 
                                              x not in fund_quant_EI_info.secID.values for x in fund_quant_info.secID]]
    # ----------------------------- 非量化基金 -----------------------
    # 指数基金
    fund_index_info = fund_info[fund_info.indexFund=='I']
    fund_index_info = fund_index_info[[x not in fund_quant_info.secID.values for x in fund_index_info.secID]]
    # 混合型基金
    fund_commingle_info = fund_info[fund_info.category=='H']
    fund_commingle_info = fund_commingle_info[[x not in fund_quant_info.secID.values and
                                               x not in fund_index_info.secID.values for x in fund_commingle_info.secID]]
    # 普通股票型基金
    fund_equ_info = fund_info[fund_info.category=='E']
    fund_equ_info = fund_equ_info[[x not in fund_quant_info.secID.values and
                                   x not in fund_index_info.secID.values and
                                   x not in fund_commingle_info.secID.values for x in fund_equ_info.secID]]
    # ----------------------------- 全部基金 -----------------------
    fund_all_info = fund_equ_info.append(fund_commingle_info).append(fund_index_info).append(fund_quant_info).drop_duplicates('secID')

    fund_secs = {'all': fund_all_info.secID.values, 'equ': fund_equ_info.secID.values, 
                 'commingle': fund_commingle_info.secID.values, 'index': fund_index_info.secID.values, 
                 'quant': fund_quant_info.secID.values, 'quant_active': fund_quant_active_info.secID.values, 
                 'quant_EI': fund_quant_EI_info.secID.values, 'quant_hedge': fund_quant_hedge_info.secID.values}
    # ----------------------------- 基金分类整理 -----------------------
    category_lv1_dict = {'equ':u'普通股票型', 'index':u'指数基金', 'commingle':u'股票混合', 
                         'quant_active':u'量化基金', 'quant_EI':u'量化基金', 'quant_hedge':u'量化基金'}
    category_lv2_dict = {'equ':u'普通股票型', 'index':u'指数基金', 'commingle':u'股票混合', 
                         'quant_active':u'量化主动', 'quant_EI':u'量化指数增强', 'quant_hedge':u'量化对冲'}

    fund_all_info = fund_all_info.set_index('secID')
    fund_all_info['categoryLevel1'] = np.NaN
    fund_all_info['categoryLevel2'] = np.NaN

    for x in fund_secs.keys():
        if x == 'all' or x == 'quant':
            continue
        fund_all_info['categoryLevel1'][fund_secs[x]] = category_lv1_dict[x]
        fund_all_info['categoryLevel2'][fund_secs[x]] = category_lv2_dict[x]    
    fund_all_info = fund_all_info.reset_index()
    # ----------------------------- 基金净值整理 -----------------------
    # 基金前复权净值
    fund_start = DataAPI.FundNavGet(dataDate=start_, field='secID,endDate,ADJUST_NAV')
    fund_end = DataAPI.FundNavGet(dataDate=end_, field='secID,endDate,ADJUST_NAV')  
    fund_all_16return = pd.merge(fund_start, fund_end, how='inner', on='secID')
    fund_all_16return['return(%)'] = np.round((fund_all_16return['ADJUST_NAV_y'] / fund_all_16return['ADJUST_NAV_x'] - 1)*100, 2)
    # 复权净值有问题的基金剔除
    bug_secID = []
    for i in fund_all_16return['secID'].tolist():
        test = DataAPI.FundNavGet(secID=i, beginDate=start_, endDate=end_)
        if abs(test['ADJUST_NAV'].pct_change().max()) > 0.1:
            bug_secID.append(i)

    fund_all_16return = fund_all_16return[~fund_all_16return['secID'].isin(bug_secID)]
    fund_all_16return = fund_all_16return[['secID','return(%)']].merge(fund_all_info, on='secID').set_index('secID')
    # 基金规模(亿)                            
    tmp_asset = DataAPI.FundAssetsGet(reportDate=report_date,secID=fund_all_16return.index.values)
    fund_all_16return['fundSize(1e8)'] = np.round(tmp_asset.set_index('secID')['totalAsset'][fund_all_16return.index] / 1e8, 2)
    # 统计不同类型的基金收益均值
    fund_type_meanret = {tp: fund_all_16return.ix[fund_secs[tp]]['return(%)'].dropna().mean()
                              for tp in fund_secs.keys()}
    # 统计不同类型的基金收益
    fund_type_ret = {
        tp: fund_all_16return.ix[fund_secs[tp]]
        [['return(%)','fundSize(1e8)','secShortName','categoryLevel2','establishDate','managerName']].
        sort_values('return(%)',ascending=False) for tp in fund_secs.keys()}

    return fund_type_ret,fund_type_meanret


# 获取基金相对基准的factor_expose
def fund_alp_expose(fundTicker, benchmarkTicker, dateList):
    
    factor_list = ['LCAP','RSTR12','REVS20','HBETA','PE','PCF','HSIGMA','CMRA','EGRO','OperatingProfitGrowRate','FEARNG','PB','MLEV','BLEV','VOL20','VOL60','ILLIQUIDITY']
    
    factor_dict = {'LCAP':['SIZE','SIZENL'], 'RSTR12':['MOMENTUM'], 'REVS20':['MOMENTUM'], 'HBETA':['BETA'], 'PE':['EARNYILD'], 'PCF':['EARNYILD'], 'HSIGMA':['RESVOL'], 'CMRA':['RESVOL'], 'EGRO':['GROWTH'], 'OperatingProfitGrowRate':['GROWTH'], 'FEARNG':['GROWTH'], 'PB':['BTOP'], 'MLEV':['LEVERAGE'], 'BLEV':['LEVERAGE'], 'VOL20':['LIQUIDTY'], 'VOL60':['LIQUIDTY'], 'ILLIQUIDITY':['LIQUIDTY']}
    
    Fund_FactorsSum = pd.DataFrame(index=dateList, columns=factor_list)
    bm_FactorsSum = pd.DataFrame(index=dateList, columns=factor_list)
    for i in range(len(dateList)):
        date_ = dateList[i]
        #取持仓公布日前的最后一个交易日
        dateDF = DataAPI.TradeCalGet(exchangeCD=u"XSHG,XSHE",beginDate=date_.replace("-",""),endDate=date_.replace("-",""),field=[u"prevTradeDate",u"isOpen"],pandas="1")
        pretradeDate = dateDF.iloc[0,0].replace("-","") if dateDF.iloc[0,1] == 0 else date_.replace("-","")

        FundHoldings = DataAPI.FundHoldingsGet(reportDate=date_.replace("-",""),ticker=fundTicker,field=["holdingSecID",'ratioInNa'],pandas="1").set_index("holdingSecID")
        FundHoldings.insert(1,'fac',0)
        
        today = Date.parseISO(date_)
        l3m_date = cal.advanceDate(today, '-3M', BizDayConvention.Preceding).toISO().replace("-","")
        benchmark = DataAPI.IdxCloseWeightGet(ticker=benchmarkTicker,beginDate=l3m_date,endDate=date_.replace("-",""),field=[u"consID",u"weight",u"effDate"],pandas="1")
        benchmark = benchmark.sort(columns=['consID', 'effDate'], ascending=[True, False])
        benchmark = benchmark.drop_duplicates(subset='consID')
        benchmark.dropna(inplace=True)
        benchmark.set_index('consID', inplace=True)
        benchmark = benchmark[benchmark['effDate']==benchmark['effDate'].ix[0]]
        del benchmark['effDate']
        benchmark.insert(1,'fac',0)

        for factor in factor_list:
            df_mv = DataAPI.MktStockFactorsOneDayGet(tradeDate=pretradeDate,secID=u"",ticker=u"",field=["secID",factor],pandas="1").set_index("secID")
            if factor == 'PE' or factor == 'PCF' or factor == 'PB' or factor == 'VOL20' or factor == 'VOL60':
                df_mv = 1.0 / df_mv 
            df_factor = df_mv[factor].dropna().to_dict()
            df_factor = pd.Series(standardize(neutralize(winsorize(df_factor),pretradeDate, exclude_style_list=factor_dict[factor])))
            fund_factor = df_factor.reindex([FundHoldings.index.values])
            
            FundHoldings['fac'] = fund_factor
            Fund_FactorsSum.ix[i,factor] = np.sum(FundHoldings['fac']*FundHoldings['ratioInNa'])/np.sum(FundHoldings['ratioInNa'])
            
            benchmark_factor = df_factor.reindex([benchmark.index.values])           
            benchmark['fac'] = benchmark_factor
            bm_FactorsSum.ix[i,factor] = np.sum(benchmark['fac']*benchmark['weight'])/np.sum(benchmark['weight'])

            alp_FactorsSum = Fund_FactorsSum - bm_FactorsSum
            df_sum = alp_FactorsSum.T
            df_sum.to_csv('fund_alp_exp.csv')
            df_sum = pd.read_csv('fund_alp_exp.csv', index_col=0)
            df_sum['num'] = pd.Series([0,1,2,3,4,4,5,5,6,6,6,7,8,8,9,9,9], index=df_sum.index)
            df_sum = df_sum.groupby(df_sum['num']).transform(np.mean)
            df_sum['name'] = pd.Series(['Size','Momentum','Reversal','Beta','Earnings Yield','Earnings Yield','Volatility','Volatility','Growth','Growth','Growth','Value','Leverage','Leverage','illiquidity','illiquidity','illiquidity'], index=df_sum.index)
            df_sum.drop_duplicates('name', inplace=True)
            df_sum.set_index('name', inplace=True)
    return df_sum
```

```python
# plot各类型股票基金的收益均值
def fund_type_meanret_plot(fund_type_meanret):
    
    with plt.xkcd():
        fig = plt.figure(figsize=(13, 4))
        ax = fig.add_subplot(111)
        font.set_size(15)
        data = pd.Series(fund_type_meanret)
        data.ix[[u'all',u'equ', u'index', u'commingle', u'quant', 
                 u'quant_EI', u'quant_active', u'quant_hedge']].plot.bar(ax=ax,width=0.4,alpha=0.25, color='#1E90FF')

        ax.set_title(u'各类型股票基金的收益均值(%)', fontproperties=font, fontsize=17)
        ax.set_xticklabels([u'全部',u'普通股票',u'指数基金',u'股票混合',u'量化类全部',u'量化指数增强',u'量化主动',
                            u'量化对  冲'],rotation=0,fontproperties=font)
        ax.grid(True)
        plt.xkcd(False)

        
# 基金间的净值对比    
def fund_cases_ret_plot(fund_cases, benchmarkTicker="000905", start_="20171231", end_="20180706"):
    
    fund_cases_df = DataAPI.FundNavGet(secID=fund_cases.index, beginDate=start_, endDate=end_)
    fund_cases_df = fund_cases_df.set_index(['secShortName','endDate'])['ADJUST_NAV'].unstack().T
    fund_cases_df[u'benchmark'] = DataAPI.MktIdxdGet(ticker=benchmarkTicker, beginDate=start_, endDate=end_).set_index('tradeDate')['closeIndex']
    fund_cases_df = fund_cases_df[fund_cases_df.index >= start_].sort_index()
    fund_cases_df = fund_cases_df / fund_cases_df.ix[0]
    
    font.set_size(13)
    ax = fund_cases_df[fund_cases.secShortName.values].plot(figsize=(15,5), color=['#1E90FF','#FF69B4','b'])
    ax.legend([x.decode('utf8') for x in fund_cases.secShortName.values], prop=font, ncol=3);
    ax = fund_cases_df[u'benchmark'].plot(figsize=(15,5), color='grey', style='--',ax=ax)
    ax.legend([x.decode('utf8') for x in fund_cases.secShortName.values]+[u'benchmark'], prop=font, ncol=4, loc=0)
    plt.grid(axis='y', ls='-')
    

# 获取基金的净值
def fundnav(fundTicker, benchmarkTicker, begin, end):
    
    quant_active_cases_df = DataAPI.FundNavGet(ticker=fundTicker, beginDate=begin, endDate=end)
    quant_active_cases_df = quant_active_cases_df.set_index(['ticker','endDate'])['ADJUST_NAV'].unstack().T
    quant_active_cases_df[u'benchmark'] = DataAPI.MktIdxdGet(ticker=benchmarkTicker, beginDate=begin, endDate=end).set_index('tradeDate')['closeIndex']
    quant_active_cases_df = quant_active_cases_df[quant_active_cases_df.index>=begin].sort_index()
    quant_active_cases_df = quant_active_cases_df / quant_active_cases_df.ix[0]

    ax = quant_active_cases_df[fundTicker].plot(figsize=(12,5), lw=2, color='#00BFFF', label='fundNav')
    ax = quant_active_cases_df[u'benchmark'].plot(figsize=(12,5), lw=2, color='grey', ax=ax, label='benchmark')
    ax.legend(fontsize=16) #loc='upper right', 
    plt.grid(axis='y', ls='-')
    
    
# 获取基金的持仓股票数量和总权重
def fund_holding(fundTicker, dateList):
    holdingDF = pd.DataFrame(index = dateList, columns = ['numofStocks'])
    for i in range(len(dateList)):
        Date = dateList[i]
        #取持仓公布日前的最后一个交易日
        dateDF = DataAPI.TradeCalGet(exchangeCD=u"XSHG,XSHE",beginDate=Date.replace("-",""),endDate=Date.replace("-",""),field=[u"prevTradeDate",u"isOpen"],pandas="1")
        tradeDate = dateDF.iloc[0,0].replace("-","") if dateDF.iloc[0,1] == 0 else Date.replace("-","")
        #取得基金持仓
        FundHoldings = DataAPI.FundHoldingsGet(reportDate=u"",secID=u"",ticker=fundTicker,beginDate=u"",endDate=u"",secType="",field=u"",pandas="1")
        FundHoldingsQ = FundHoldings[FundHoldings['reportDate']==Date].reset_index(drop=True)
        #c持仓股票数量和总权重
        holdingDF.ix[i,'numofStocks'] = FundHoldingsQ.shape[0]
        holdingDF.ix[i,'weightofStocks'] = np.sum(FundHoldingsQ['ratioInNa'])

    with plt.xkcd():
        fig = plt.figure(figsize=(13,4))
        font.set_size(15)
        ax = fig.add_subplot(121)
        ax = holdingDF["numofStocks"].T.plot.bar(ax=ax, alpha=0.25, color='#1E90FF', width=0.3)
        ax.set_xticklabels(holdingDF.index,rotation=0,fontproperties=font);
        ax.set_title(u'公布持仓股票数量（只）',fontproperties=font);
        ax.grid(True)
        ax = fig.add_subplot(122)
        ax = holdingDF["weightofStocks"].T.plot.bar(ax=ax, alpha=0.25, color='#1E90FF', width=0.3)
        ax.set_xticklabels(holdingDF.index,rotation=0,fontproperties=font);
        ax.set_title(u'公布持仓股票权重（%）',fontproperties=font);
        ax.grid(True)
    

# 获取基金的指数exopse
def fund_idx_expose(fundTicker, dateList):
    
    with plt.xkcd():
        fig = plt.figure(figsize=(11,7))
        fig.set_tight_layout(True)
    
        for i in range(len(dateList)):
            Date = dateList[i]
            idx_300 = DataAPI.IdxConsGet(ticker='000300',intoDate=Date,field='',pandas="1")['consTickerSymbol'] 
            idx_500 = DataAPI.IdxConsGet(ticker='000905',intoDate=Date,field='',pandas="1")['consTickerSymbol'] 
            idx_1000 = DataAPI.IdxConsGet(ticker='000852',intoDate=Date,field='',pandas="1")['consTickerSymbol']
            idx_outside = DataAPI.IdxConsGet(ticker='000985',intoDate=Date,field='',pandas="1")['consTickerSymbol']
            FundHoldings = DataAPI.FundHoldingsGet(reportDate=Date,ticker=fundTicker,field="holdingSecID",pandas="1")
            cur_positions = sec2ticker(list(FundHoldings['holdingSecID'].values))

            idx_list = []
            for stk in cur_positions:
                if stk in list(set(idx_300)):
                    idx_list.append('idx_300')
                if stk in list(set(idx_500)):
                    idx_list.append('idx_500')
                if stk in list(set(idx_1000)):
                    idx_list.append('idx_1000')
                if stk in set(idx_outside).difference(set(idx_300), set(idx_500), set(idx_1000)): 
                    idx_list.append('idx_out')       
            universe_ind_df = pd.DataFrame(idx_list)
            ind_percent = universe_ind_df[0].value_counts() / len(universe_ind_df)

            ax = fig.add_subplot(2,2,i+1)
            ax = ind_percent.plot.barh(ax=ax, alpha=0.25, color='#1E90FF', width=0.6)
            #ind_percent.plot(kind='barh',ax=ax,alpha=0.25,color='#1E90FF',width=0.6)

            def myfunc(x, pos=0): 
                return "{0:.2f}%".format(x * 100)
            ax.set_yticklabels([item.decode('utf-8') for item in ind_percent.index],fontproperties=font,fontsize=15)
            ax.spines['top'].set_color('none')
            ax.xaxis.set_ticks_position('bottom')
            ax.set_title(Date, fontproperties=font, fontsize=15)
            ax.grid(True)    

            
# 基金相对基准的facotr_expose    
def fund_expose_plot(df_sum, yticks=[-2,-1,0,1,2]):
    
    plt.style.use('ggplot')
    fig=plt.figure(figsize=(10,10))
    fig.set_tight_layout(True)
    font.set_size(15)
    ax1=fig.add_subplot('221',projection='polar')
    ax2=fig.add_subplot('222',projection='polar')
    ax3=fig.add_subplot('223',projection='polar')
    ax4=fig.add_subplot('224',projection='polar')

    feature_label=['Size','Momentum','Reversal','Beta','EarningsYield','Volatility',
                   'Growth','Value','Leverage','illiquidity']
    fac_data = {}
    for x in range(df_sum.columns.size):
        df_dict = {df_sum.icol(x).name: df_sum.icol(x).values}
        fac_data.update(df_dict)
    theta=np.linspace(0,2*np.pi,10,endpoint=False)
    theta=np.append(theta,theta[0])

    fac_data[fac_data.keys()[0]]=np.append(fac_data[fac_data.keys()[0]],fac_data[fac_data.keys()[0]][0])
    fac_data[fac_data.keys()[0]]
    ax1.plot(theta,fac_data[fac_data.keys()[0]],'r',linewidth=2)
    ax1.fill(theta,fac_data[fac_data.keys()[0]],'r',alpha=0.3)
    ax1.set_xticks(theta)
    ax1.set_xticklabels(feature_label,y=0.05,fontproperties=font)
    ax1.set_yticks(yticks)
    ax1.set_title(fac_data.keys()[0],color='r',size=20)

    fac_data[fac_data.keys()[1]]=np.append(fac_data[fac_data.keys()[1]],fac_data[fac_data.keys()[1]][0])
    fac_data[fac_data.keys()[1]]
    ax2.plot(theta,fac_data[fac_data.keys()[1]],'g',linewidth=2)
    ax2.fill(theta,fac_data[fac_data.keys()[1]],'g',alpha=0.3)
    ax2.set_xticks(theta)
    ax2.set_xticklabels(feature_label,y=0.05,fontproperties=font)
    ax2.set_yticks(yticks)
    ax2.set_title(fac_data.keys()[1],color='g',size=20)

    fac_data[fac_data.keys()[2]]=np.append(fac_data[fac_data.keys()[2]],fac_data[fac_data.keys()[2]][0])
    fac_data[fac_data.keys()[2]]
    ax3.plot(theta,fac_data[fac_data.keys()[2]],'b',linewidth=2)
    ax3.fill(theta,fac_data[fac_data.keys()[2]],'b',alpha=0.3)
    ax3.set_xticks(theta)
    ax3.set_xticklabels(feature_label,y=0.05,fontproperties=font)
    ax3.set_yticks(yticks)
    ax3.set_title(fac_data.keys()[2],color='b',size=20)

    fac_data[fac_data.keys()[3]]=np.append(fac_data[fac_data.keys()[3]],fac_data[fac_data.keys()[3]][0])
    fac_data[fac_data.keys()[3]]
    ax4.plot(theta,fac_data[fac_data.keys()[3]],'y',linewidth=2)
    ax4.fill(theta,fac_data[fac_data.keys()[3]],'y',alpha=0.3)
    ax4.set_xticks(theta)
    ax4.set_xticklabels(feature_label,y=0.05,fontproperties=font)
    ax4.set_yticks(yticks)
    ax4.set_title(fac_data.keys()[3],color='y',size=20)
    plt.show()      
```
