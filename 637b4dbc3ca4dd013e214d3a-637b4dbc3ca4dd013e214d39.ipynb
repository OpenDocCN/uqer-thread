# <font color=#0099ff>二、 功能支持 </font>
---
## 1 Python3编程环境支持场外基金回测
- python编程环境下，新增场外基金的回测，而且此回测过程更贴合基金的交易规则，比如下单形式有两种选择，包括申赎下单（by_order）或`按照权重及时进行持仓分配下单(by_weight)`，尤其是by_weight，更符合基金策略的研究角度。其他特点还有根据金额进行申购，根据份额进行赎回；对基金的申购赎回订单进行模拟确认；包括现金分红及分红再投的两种分红模式等。
- 网页端优矿只提供python2的编程环境，此功能适用于使用终端的客户，详情参见帮助文档手册。

## 2 回测投资域加入科创50成分股
- 增加投资域选项，更方便客户做策略回测与研究。
- 该功能适用于任何客户，详情参见帮助文档手册。
## 3 中性化函数mfhandler支持使用2014版或2021版申万行业
- 申万行业研究所在2021年11月发布了新的申万行业分类：`一级行业`从28个调整到`31个`，二级行业从104个调整到134个，三级行业从227个调整到346个，而由于mfhanlder在对因子进行中性化时涉及到了股票的行业分类，所以对此函数进行了迭代，新增neutralize_pit函数。该函数用法是：在2021年10月前中性化时，使用的是14版申万行业数据；在21年10月份之后中性化时，使用申万21版行业数据。
- 该功能适用于任何客户，详情参见帮助文档手册。
## 4 quartz回测过程可以获取当前组合截止到上一交易日的净值数据等
- 在策略运行过程中，有必要获取当前策略截止到上一交易日的净值数据/回撤数据/收益波动率数据，以方便策略在盘中的自我检测或作为判断条件进入下一步的操作，所以新增函数`context.get_nav`、`context.get_max_drawdown`与`context.get_volatility`。
- 该功能适用于任何客户，详情参见帮助文档手册。
## 5 接口试用次数提升
- 考虑到用户在试用时充分测试的场景，对试用次数规则进行了调整。变更之前的限制为“单接口总调用次数1000次”，现调整为`单接口调用次数每日1000次`，`调用次数每天0点重置`，更好地满足用户的接口试用需求。
- 该功能适用于任何使用网页端或终端的试用客户。

- 1.1 新增接口查看筛选栏，可以从“全部接口”、“权限内接口”、“权限外接口”三个分类进行接口筛选。其中，“全部接口”表示优矿所有接口一览，“权限内接口”表示用户当前版本可调取的接口，“权限外接口”表示需要额外购买的接口。
![图片注释](http://storage-uqer.datayes.com/564aee87f9f06c4446b4829b/3a8c7ba4-6984-11ed-93bd-0242ac140002)
- 1.2 新增接口版本标签，版本分为“极速版”、“标准版”和“旗舰版”三个选项。灰色表示该版本无此接口；蓝色表示该版本包含此接口。例如：DataAPI.BondGet，显示为标准版和旗舰版可调取，而极速版未包含。
![图片注释](http://storage-uqer.datayes.com/564aee87f9f06c4446b4829b/3aaf32fc-6984-11ed-93bd-0242ac140002)
