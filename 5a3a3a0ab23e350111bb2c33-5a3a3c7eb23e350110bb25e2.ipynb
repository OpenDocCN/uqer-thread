# 次新股分析

次新股一般是指上市一年内的个股，次新股的投资逻辑：
a、IPO价格抑制，市盈率折价，高成长性，具有较好的上涨空间
b、新股流通盘小，资本运作空间大，资金可挖掘的热点多
c、最重要的，上方无套牢盘，且由于大小非限制，无砸盘压力
d、筹码分散，机构持仓比低，便于游资收集筹码

次新股在上市之初，连续开板之后会有资金获利离场，等到短期获利者完全出货，又会面临一段长期上涨的行情，因而次新股的走势一般是N字型走势，可根据部分技术指标选择买卖点

通常的技术面买卖时间点如下：
a.上市买入
b.开板买入（开板后第二天，涨幅低于9%）
c.上市后20日买入（上市后20个交易日，涨幅低于9%）
d.股价突破20日均线买入（收盘价高于20日均线，第二天，涨幅低于9%）
e.股价形成金叉后买入（5日均线上穿20日均线，第二天，涨幅低于9%）
f.换手率跌破阈值后买入（低于9%后第二天，涨幅低于9%）

次新股的炒作属于资金驱动，考虑到主题性资金快进快出，取5日10日均线金叉死策略，这种简单策略能跑赢大盘的话，就有潜力做更多挖掘。

在上面的策略中，买入顺序，是按照IPO最近时间排列的，是否有更好的选股准则？

以下我们建构多因子策略：
因子集合：PE、流通市值、对数市值、净利润增速、20alpha，20日换手率、5日换手率
显然资金炒作，要炒作绩效好、流通股少的股票，游资快进快出的特点，也决定了交易量等技术指标非常重要。

在对技术指标进行中心化、去极值等操作后，对因子加权，构成组合指标，在换仓的时候，先买入综合权重大的股票。
金叉买入，死叉卖出，周度调仓，最多持股30只

这里我们的因子权重是固定的，如何动态的调整权重，自适应不同的市场状态，就是后续值得改进的地方。
