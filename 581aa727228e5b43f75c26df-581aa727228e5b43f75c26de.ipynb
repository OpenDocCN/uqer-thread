```python
"""
fundamental Analysis Factors
--------------------------
"""
from __future__ import division
import numpy  as np
import pandas as pd

from CAL.PyCAL import *
cal = Calendar('China.SSE')
```

```python
from functools import wraps
def wrapper_func_cache(func):
    @wraps(func)
    def wrapped(*args, **kwargs):
        if not hasattr(func, '__cache_result'):
            func.__cache_result = {}
        
        hash_key = [str(i) for i in args] + [str(j) for j in kwargs.iteritems()]
        cache_key = hash(tuple(hash_key))
        
        if cache_key not in func.__cache_result:
            func.__cache_result[cache_key] = func(*args, **kwargs)
            
        return func.__cache_result[cache_key]
    return wrapped

@wrapper_func_cache
def load_financial_statement(securities,start,end,balance_fields=None,cashflow_fields=None,income_fields=None):
    """
    securities = ['600017.XSHG', '601008.XSHG']
    balance_fields = ['TLiab','TShEquity']
    cashflow_fields = ['CPaidInvest','NCEEndBal']
    income_fields = ['operateProfit','finanExp']
    """
    TradeCal = DataAPI.TradeCalGet(exchangeCD=u"XSHG,XSHE",beginDate=start,endDate=end,field="calendarDate,isOpen",pandas="1")
    TradeCal = list(TradeCal[TradeCal['isOpen'] == 1]["calendarDate"])
    datetimes = sorted(list(set(TradeCal)))
    # Iterate over all tradable days
    fundamental = {}
    for current_date in datetimes:
        window_length = '-6M'
        start_date = cal.advanceDate(current_date,window_length,BizDayConvention.Preceding).strftime('%Y%m%d')
        tmpStatement = pd.DataFrame() 
        
        # 合并资产负债表The balance sheet
        if balance_fields is not None:
            balance = pd.DataFrame()
            for rpt_type in ['Q1', 'S1', 'Q3', 'A']:
                try:
                    tmp  = DataAPI.FdmtBSGet(secID=securities,beginDate=start_date,field=['secID', 'endDate', 'publishDate']+balance_fields, publishDateEnd=current_date, reportType=rpt_type,pandas="1")
                except:
                    tmp = pd.DataFrame()
                balance = pd.concat([balance, tmp], axis=0)

            balance = balance.sort(columns=['secID', 'endDate', 'publishDate'], ascending=[True, False, False])
            balance = balance.dropna()
            balance = balance.drop_duplicates(subset='secID')
            balance.pop('publishDate')
            balance.pop('endDate')
            balance = balance.set_index('secID')
            tmpStatement = pd.concat([tmpStatement, balance], axis=1)

        # 合并现金流量表The cash flow statement
        if cashflow_fields is not None:
            cashflow = pd.DataFrame()
            for rpt_type in ['Q1', 'S1', 'Q3', 'A']:
                try:
                    tmp  = DataAPI.FdmtCFGet(secID=securities,beginDate=start_date,field=['secID', 'endDate', 'publishDate']+cashflow_fields, publishDateEnd=current_date, reportType=rpt_type,pandas="1")
                except:
                    tmp = pd.DataFrame()
                cashflow = pd.concat([cashflow, tmp], axis=0)

            cashflow = cashflow.sort(columns=['secID', 'endDate', 'publishDate'], ascending=[True, False, False])
            cashflow = cashflow.dropna()
            cashflow = cashflow.drop_duplicates(subset='secID')
            cashflow.pop('publishDate')
            cashflow.pop('endDate')
            cashflow = cashflow.set_index('secID')
            tmpStatement = pd.concat([tmpStatement, cashflow], axis=1)

        # 合并利润表The income statement
        if income_fields is not None:
            income = pd.DataFrame()
            for rpt_type in ['Q1', 'S1', 'Q3', 'A']:
                try:
                    tmp  = DataAPI.FdmtISGet(secID=securities,beginDate=start_date,field=['secID', 'endDate', 'publishDate']+income_fields, publishDateEnd=current_date, reportType=rpt_type,pandas="1")
                except:
                    tmp = pd.DataFrame()
                income = pd.concat([income, tmp], axis=0)

            income = income.sort(columns=['secID', 'endDate', 'publishDate'], ascending=[True, False, False])
            income = income.dropna()
            income = income.drop_duplicates(subset='secID')
            income.pop('publishDate')
            income.pop('endDate')
            income = income.set_index('secID')
            tmpStatement = pd.concat([tmpStatement, income], axis=1)
        fundamental[current_date] = tmpStatement
    return pd.Panel.from_dict(fundamental)
```

# PROFITABILITY 盈利性因子

```python
def ROIC(securities,start,end):
    """
    资本回报率（ROIC）是用来评估一个公司或其事业部门历史绩效的指标。贴现现金流,正如我们所知,它决定着任何公司的最终（未来）价值,它也是对公司进行评估的一个最主要的指标。
    """ 
    #IC：投入资本=总资产-无息负债-超额现金=股东权益+有息负债-超额现金
    #期初有息负债=期初短期借款(STBorr)+期初长期借款(LTBorr)+期初应付债券(bondPayable)+期初一年之内到期的非流动负债(NCLWithin1Y)+期初少数股东权益(minorityInt)
    #至于超额现金，这也是一个估计数字，取决于对公司经营特点的深入了解，就A股大部分公司而言，我们可以认为它们没有任何超额现金，但像茅台这样的公司则较特殊。
    #资产负债表：TShEquity（所有者权益合计）、LTBorr（长期借款）、minorityInt（少数股东权益）
    balance_fields = ['TShEquity','LTBorr','minorityInt']
    # 现金流量表
    # cashflow_fields = ['CPaidInvest','NCEEndBal']

    #利润表：# 税后净营业利润（NOPAT）= （operateProfit （营业利润）- finanExp（财务费用））*(1-经营利润所得税（33%）)
    income_fields = ['operateProfit','finanExp'] 
    # 调用合并财务报表数据
    fs_data = load_financial_statement(securities,start,end,balance_fields,None,income_fields)
    fs_data = fs_data.swapaxes('items','major')

    return ((fs_data.minor_xs('operateProfit') - fs_data.minor_xs('finanExp')) * (1-0.33)) / (fs_data.minor_xs('TShEquity') + fs_data.minor_xs('LTBorr') + fs_data.minor_xs('minorityInt'))
```
