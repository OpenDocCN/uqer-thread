```python
#方案三：对买入股票池按前X涨幅排序
lastDay = '20181015'
data_start = cal.advanceDate(lastDay,'-5Y',BizDayConvention.Preceding).strftime('%Y%m%d')
# 排除一年内上市新股
listDate = cal.advanceDate(lastDay,'-1Y',BizDayConvention.Preceding).strftime('%Y%m%d')
MktEqud_df = DataAPI.MktEqudGet(tradeDate=listDate,field=[u"secID"],pandas="1")
MktEqud_df.set_index("secID",inplace=True)
base_universe = [sec for sec  in MktEqud_df.index ]
#去除当日未开盘股
MktEqud_df = DataAPI.MktEqudGet(tradeDate=lastDay,secID=base_universe,field=[u"secID",u"ticker",u"secShortName"],isOpen=1,pandas="1")
MktEqud_df.set_index("secID",inplace=True)
# 排除ST S * '退'
MktEqud_df = MktEqud_df[~MktEqud_df.secShortName.str.contains('S')].copy()

# 排除B股
MktEqud_df = MktEqud_df[MktEqud_df.ticker<'900000'].copy()
tickers_listDate = DataAPI.SecIDGet(assetClass=u"E",ticker=list(MktEqud_df.ticker.values),field=u"secID,listDate",pandas="1")
tickers_listDate = tickers_listDate.set_index("secID")
base_universe = [sec for sec in tickers_listDate.index]
from lib.data_loader import load_market_data

market_data = load_market_data(base_universe,data_start,lastDay,dependencies=['Close'])
Close_data = market_data.minor_xs('Close')
close_df = DataAPI.MktEqudGet(tradeDate=lastDay,secID=base_universe,field=[u"secID",u"closePrice"],isOpen=1,pandas="1")
close_df.set_index("secID",inplace=True)
close_df['maxPrice'] = Close_data.max()
close_df['pct_change'] = 1-close_df['closePrice']/close_df['maxPrice']
close_df.sort_values(by=['pct_change'],ascending=False, inplace=True)
```
