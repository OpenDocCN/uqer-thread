由于股票市场的非线性特性和大量异常事件的存在，传统线性模型对市场的拟合效果并不好，本文考虑到市场不同状态下的不同特征，试图采用分阶段的方法先将市场划分成不同状态在进行训练拟合。由于时间有限，本文仅给出一个研究方法，供学术参考。<br>
这一方法主要由四个部分组成：

> 1，采用混沌分析理论，计算出时间序列的关联维数，从而确定神经网络的输入维度。
> 
> 2，计算时间序列的转折点，从而将时间序列分成若干种状态。
> 
> 3，将新输入值归类到某一状态。
> 
> 4，根据该状态的神经网络得到输出值。

## **1. Correlation dimension**

这里我们采用混沌分析的方法，基于内生维数计算时间序列的关联维数。相比GARCH类模型，混沌分析计算维数的鲁棒性更好，即更加稳定。关联维数的计算方法如下,对于序列${s_t:t=1,...T}$：
$$D^m=lim_{n→0}\frac{ln(C_m(r))}{ln(r)}$$

其中：
$$C(r)=lim_{T→∞}\frac{1}{T^2}∑_{1<i,j<T}I_r(s_i,s_j)$$

I为示性函数，等于$||x-y||<r$,或者0，即统计在r半径内的点，m为内生维数，由外界给定。

```python
#关联维数计算
def corr_dim(data,r):
    count=0
    for i in data.columns:
        for j in data.columns:
            a=((data[i]-data[j])**2).sum()
            if a<r:
                count=count+1
    K=[np.log(r),np.log(count/(len(data)+0.00)**2)]
    return K
```

## **2. Pettitt test**

为了将时间序列分为多个阶段，我希望采用转折点分类的方法，这种方法的优点在于分类的同时能保持时间序列的完整性。为了寻找转折点我采用了Pettitt检验，寻找转折点K的方法如下:
$$K_T=max_{t∈[1,T)}|U_t,T|$$

其中：
$$U_{t,T}=∑_{j=1}^{t}∑_{i=t+1}^{T}sgn(X_i-X_j)$$

K的分布(P值)为：
$$P=2e^{\frac{-6K^2}{T^2+T^3}}$$

```python
#pettitt检验
#-----------------------------------------------------------------------------
def pettitt_test(U):
    T=len(U)
    P=pd.Series(range(T))
    for t in (U.index)[1:-1]:
        for i in (U.index)[:t+1]:
            for j in (U.index)[t+1:]:
                if U[i]>U[j]:
                    P[t]=P[t]+1
                elif U[i]<U[j]:
                    P[t]=P[t]-1
                else:
                    P[t]=P[t]
    K=max(U)
    p=2*np.exp(-6*(K**2)/(T**3+T**2))
    return K,p


#搜索转折点
#-----------------------------------------------------------------------------
def change_point(U,count):
    t=0
    K,p=pettitt_test(U)
    t=list(U).index(K)
    if p<0.1 and t!=0 and t!=len(U)-1:
        print K
        count=count+1
        count=change_point(U[:t+1],count)
        a=U[t+1:]
        a.index=range(len(U[t+1:]))
        count=change_point(a,count)
    return count
```
