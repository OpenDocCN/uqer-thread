# 【大师系列】詹姆斯.奥肖内西价值投资法

詹姆斯.奥肖内西一直被公认为美国重要的金融专家和股票数量分析领域的先驱。福布斯网站（Forbes.com）把詹姆斯.奥肖内西与本杰明.格雷厄姆，沃伦.巴菲特和彼得.林奇一起称为“传奇的投资者”（Legendary Investors）。詹姆斯.奥肖内西在其经典之作《华尔街股市投资经典》中指出，在大多数情况下，被动、机械化的投资系统会打败由人组成的投资系统，即使是基金经理也不例外。
--------
***
参考申万研报《詹姆斯.奥肖内西价值投资法——申万大师系列_价值投资篇之四》，具体策略与量化实现有下面六条。

>1.剔除公共事业股票和暂停交易的股票。
>
>2.以1 的结果为样本，筛选出总市值大于平均市值的股票。
>
>3.以2 的结果为样本，剔除每股现金流量小于0 的股票，再筛选出股价现金流量比小于平均值的股票。
>
>4.以3 的结果为样本，筛选出总股本大于平均值的股票。(该条经测试，非常不适合中国市场，予以去掉)
>
>5.以4 的结果为样本，剔除每股营收小于0 的股票，再筛选出股价营收比小于平均值2/3 的股票。
>
>6.以5 的结果为样本，筛选出股利收益率最高的30 只股票。

```python
#行业分类与股票相对市值一般变动不大，我们事先计算出满足一二条件的股票池，减少运行时间。
from CAL.PyCAL import *
import numpy as np
import pandas as pd
from pandas import DataFrame
df_ind = DataAPI.EquIndustryGet(industryVersionCD="010303",intoDate='20150910', field=['secID','industryName1'])
filter_universe =  list(set(df_ind['secID']) - set(df_ind[df_ind['industryName1']=='公用事业']['secID']))
setUniverse = set_universe('A')
filter_universe = list(set(filter_universe) & set(setUniverse ))

eq_marketValue = pd.DataFrame()
for s in xrange(300,len(filter_universe),300): #由于API的读取限制，分批运行API。
    tmp = DataAPI.MktEqudGet(secID=filter_universe[s-300:s],tradeDate=u"20150914",field=['secID','marketValue'],pandas="1")
    eq_marketValue = pd.concat([eq_marketValue,tmp],axis = 0)
tmp = DataAPI.MktEqudGet(secID=filter_universe[s:],tradeDate=u"20150914",field=['secID','marketValue'],pandas="1")
eq_marketValue = pd.concat([eq_marketValue,tmp],axis = 0)   
eq_marketValue = eq_marketValue[eq_marketValue['marketValue'] > np.mean(eq_marketValue['marketValue'])]
fliter2_universe = eq_marketValue['secID'].tolist()
print len(fliter2_universe),fliter2_universe
```

```python
#每股现金流量是一个TTM值，计算公式是营业业务所带来的净现金流量/总股数，所以股价现金流量比即为总市值/营业业务所带来的净现金流量
def ncfOperateA(account):
    today = account.current_date
    cal = Calendar('China.SSE')
    lastDay = cal.advanceDate(today,'-1B',BizDayConvention.Unadjusted)
    today_str = lastDay.strftime("%Y%m%d")
    start_date = cal.advanceDate(lastDay,'-500B',BizDayConvention.Unadjusted)
    start_date_str = start_date.strftime("%Y%m%d")
    
    equity = pd.DataFrame()
    for rpt_type in ['Q1','S1','A']:
        try:
            tmp = DataAPI.FdmtCFGet(secID=account.universe, field=['secID', 'endDate', 'publishDate', 'NCFOperateA'], beginDate=start_date_str, publishDateEnd=today_str,  reportType=rpt_type)
        
        except:
            tmp = DataFrame()
        equity = pd.concat([equity, tmp], axis=0)

    equity = equity.sort(columns=['secID','endDate','NCFOperateA'], ascending=[True, False, False])
    equity = equity.drop_duplicates(cols=['secID','endDate'])
    equity = equity.dropna()
    
    signalNCFO = {}
    for stock in account.universe:
        try:
            num = len(equity[equity['secID']==stock]['endDate'])
            if (num>=4):
                endDate = equity[equity['secID']==stock]['endDate'].iloc[0]
                month= int(endDate[6])
                if (month==2) :
                    NCFO_TTM = equity[equity['secID']==stock]['NCFOperateA'].iloc[0]
                elif  (month==6):
                    NCFO_TTM = equity[equity['secID']==stock]['NCFOperateA'].iloc[0]-equity[equity['secID']==stock]['NCFOperateA'].iloc[3]+equity[equity['secID']==stock]['NCFOperateA'].iloc[2] 
                elif (month==3):
                    NCFO_TTM = equity[equity['secID']==stock]['NCFOperateA'].iloc[0]-equity[equity['secID']==stock]['NCFOperateA'].iloc[3]+equity[equity['secID']==stock]['NCFOperateA'].iloc[1]
                else:
                    NCFO_TTM = 'NA'
                if len(signalNCFO) == 0:
                    signalNCFO['secID'] = [stock]
                    signalNCFO['signalNCFO'] = [NCFO_TTM]
                else:
                    signalNCFO['secID'].append(stock)
                    signalNCFO['signalNCFO'].append(NCFO_TTM)    
        except:
            pass
 
    signalNCFO = DataFrame(signalNCFO)
    filterList1 = signalNCFO['secID'].tolist()
    signalNCFO =  signalNCFO[signalNCFO['signalNCFO'] > 0] 
    
    filter1_NCFO = {}
    for stock in filterList1:
        try:
            tmp = DataAPI.MktEqudGet(secID=stock,tradeDate=today_str,field='marketValue',pandas="1")
            price_NCFO = tmp['marketValue'].values[0]/signalNCFO[signalNCFO['secID'] == stock]['signalNCFO'].values[0]

            if len(filter1_NCFO) == 0:
                filter1_NCFO['secID'] = [stock]
                filter1_NCFO['price_NCFO'] = [price_NCFO]
            else:
                filter1_NCFO['secID'].append(stock)
                filter1_NCFO['price_NCFO'].append(price_NCFO)    
        except:
            pass

    filter1_NCFO = DataFrame(filter1_NCFO)
    
    filter1_NCFO = filter1_NCFO[filter1_NCFO['price_NCFO'] < np.mean(filter1_NCFO['price_NCFO'])]
    filterList2 = filter1_NCFO['secID'].tolist()
    return filterList2
```

```python
#计算每股营收TTM值，并拿来筛选
def filter_Rev(account,filter_list):
    today = account.current_date
    cal = Calendar('China.SSE')
    lastDay = cal.advanceDate(today,'-1B',BizDayConvention.Unadjusted)
    today_str = lastDay.strftime("%Y%m%d")
    start_date = cal.advanceDate(lastDay,'-500B',BizDayConvention.Unadjusted)
    start_date_str = start_date.strftime("%Y%m%d")
    
    equity = pd.DataFrame()
    
    for rpt_type in ['Q1','S1','A']:
        try:
            tmp = DataAPI.FdmtISGet(secID=filter_list, field=['secID', 'endDate', 'publishDate', 'tRevenue'], beginDate=start_date_str, publishDateEnd=today_str,  reportType=rpt_type)
        
        except:
            tmp = DataFrame()
        equity = pd.concat([equity, tmp], axis=0)
    equity = equity.sort(columns=['secID','endDate','tRevenue'], ascending=[True, False, False])
    equity = equity.drop_duplicates(cols=['secID','endDate'])
    equity = equity.dropna()
    
    signalRev = {}
    for stock in filter_list:
        try:
            num = len(equity[equity['secID']==stock]['endDate'])
            if (num>=4):
                endDate = equity[equity['secID']==stock]['endDate'].iloc[0]
                month= int(endDate[6])
                if (month==2) :
                    Rev_TTM = equity[equity['secID']==stock]['tRevenue'].iloc[0]
                elif  (month==6):
                    Rev_TTM = equity[equity['secID']==stock]['tRevenue'].iloc[0]-equity[equity['secID']==stock]['tRevenue'].iloc[3]+equity[equity['secID']==stock]['tRevenue'].iloc[2] 
                elif (month==3):
                    Rev_TTM = equity[equity['secID']==stock]['tRevenue'].iloc[0]-equity[equity['secID']==stock]['tRevenue'].iloc[3]+equity[equity['secID']==stock]['tRevenue'].iloc[1]
                else:
                    Rev_TTM = 'NA'
                if len(signalRev) == 0:
                    signalRev['secID'] = [stock]
                    signalRev['signalRev'] = [Rev_TTM]
                else:
                    signalRev['secID'].append(stock)
                    signalRev['signalRev'].append(Rev_TTM)    
        except:
            pass
 
    signalRev = DataFrame(signalRev)
    signalRev =  signalRev[signalRev['signalRev'] > 0] 
    filterList5 = signalRev['secID'].tolist()

    filter5_Rev = {}
    for stk in filterList5:
        if  account.referencePrice[stk] > 0:
            tmp = DataAPI.MktEqudGet(secID=stk,tradeDate=today_str,field='marketValue',pandas="1")
            price_Rev = tmp['marketValue'].values[0]/signalRev[signalRev['secID'] == stk]['signalRev'].values[0]
            if len(filter5_Rev) == 0:
                filter5_Rev['secID'] = [stk]
                filter5_Rev['signalRev'] = [price_Rev]
            else:
                filter5_Rev['secID'].append(stk)
                filter5_Rev['signalRev'].append(price_Rev) 
    
    filter5_Rev = DataFrame(filter5_Rev)
    
    filter5_Rev = filter5_Rev[filter5_Rev['signalRev'] < np.mean(filter5_Rev['signalRev'])*2/3]
    filterList6 = filter5_Rev['secID'].tolist()
    return filterList6
```

```python
#计算过去一年股利收益率
def dividendYield(account,filter_Div):
    today = account.current_date
    cal = Calendar('China.SSE')
    lastDay = cal.advanceDate(today,'-1B',BizDayConvention.Unadjusted)
    today_str = lastDay.strftime("%Y%m%d")
    start_date = cal.advanceDate(lastDay,'-366D',BizDayConvention.Unadjusted)
    start_date_str = start_date.strftime("%Y%m%d")
    
    eq_Div = DataAPI.EquDivGet(secID=filter_Div,beginDate = start_date_str,endDate = today_str,field=['perCashDiv','secID','exDivDate'],pandas="1")
    
    eq_closePrice = pd.DataFrame()
    for stk in eq_Div['secID'].tolist():
        
        try:
            trade_date = eq_Div[eq_Div['secID']==stk]['exDivDate'].values[0]
            trade_date = Date.strptime(trade_date,"%Y-%m-%d")
            trade_date = Date.strftime(trade_date,"%Y%m%d")

            tmp = DataAPI.MktEqudGet(secID=stk,tradeDate=trade_date,field=['secID','closePrice'],pandas="1")
            eq_closePrice = pd.concat([eq_closePrice,tmp],axis=0)
        except:
            pass
        
    eq_closePrice.set_index('secID',inplace=True)
    eq_closePrice = eq_closePrice['closePrice'].to_dict()

    eq_Div.set_index('secID',inplace=True)

    eq_Div['price'] = np.nan

    for key in eq_closePrice:
        eq_Div['price'][key] = eq_closePrice[key]
    eq_Div['divYield'] = eq_Div['perCashDiv']/eq_Div['price']

    eq_Div = eq_Div[~np.isnan(eq_Div['divYield'])]
    filter_list = eq_Div.sort(columns = 'divYield',ascending=False)
    buylist = list(filter_list.index)
    return buylist[0:30] 
```
