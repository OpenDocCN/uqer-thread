一、CCI简介
***
CCI是常用的技术指标之一，属于超买超卖型指标。通常情况下CCI会运行于[-100, 100]， CCI小于-100时进入超卖区，大于100时进入超买区。CCI的基本操作是CCI上穿100时买入，下穿100卖出；下穿-100卖出，上穿-100买入。

二、策略简介
***
简单实现了机械化地开仓/平仓操作，对日期参数的选择和开仓/平仓标准进行了适当优化，不再是教条化的-100，100。教条化执行会造成过早进场或者离场。 
以上证50指数为Benchmark，交易标的选用50ETF（510050）。经过简单的测试，时间参数选用21，开仓条件上穿80，平仓条件为下穿20.
时间参数过小CCI波动过大，虚假信号过多。在下跌趋势中，一个小级别反弹，就可能出发买入信号，造成过早入场；反之在上涨趋势中，一个小级别回调，可能发出卖出信号，造成过早离场。
开仓和平仓线制定也没有完全遵照理论来。在趋势不明显的震荡行情中，CCI可能刚刚上穿100就回落，反之刚刚下穿-100就反弹，形成反复揉搓，造成很大损失。在这种行情中一般采取上穿-100开仓，下穿100平仓。开仓条件设为上穿80，平仓条件设为下穿20的意图是在尽早确定标的上涨趋势开仓，当上涨趋势破坏后及早平仓。

三、不足与需要改进之处
***
1、简单一只标的，没有采用投资组合。不过本文的目的就是focus on CCI，投资组合暂不考虑。
2、仓位管理是粗暴的全仓进全仓出。以后需要适当考虑仓位管理
3、判断开仓平仓条件过于简单。可以考虑其他指标配合，如OBV，KDJ等。如果仅从CCI的角度，可以考虑CCI Correction策略或者加入CCI背离的判断。
