```python
#20181012#赵作伦，可实习。
#243
#加载程序包
import datetime
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn import preprocessing
import matplotlib.pyplot as plt
import seaborn
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.feature_selection import SelectFromModel
#第一步：妥协于优矿的API，为获得交易时间列表trade_time,我们选取随机一只股票的数据，x和y进行结合后选取的都有数据的日期
data_timex=DataAPI.MktStockFactorsDateRangeGet(secID=u"",ticker=u"600376",beginDate=u"20161011",endDate=u"20181011",field=u"",pandas="1")
data_timey=DataAPI.EquRetudGet(secID=u"",ticker=u"600376",listStatusCD=u"",
beginDate=u"20161011",endDate=u"20181011",dailyReturnNoReinvLower=u"",dailyReturnNoReinvUpper=u"",dailyReturnReinvLower=u"",dailyReturnReinvUpper=u"",isChgPctl=u"",field=u"",pandas="1")#提取数据
data_timex.dropna(axis=0,inplace=True,how='all')#hang
data_timex.dropna(axis=1,inplace=True,how='any')#列指标删除
data_timex.set_index('tradeDate',inplace=True)
data_timey.set_index('tradeDate',inplace=True)#接下来进行数据合并，取公共有数据的股票
data_timey.dropna(axis=0,how='all')
data_timey=data_timey['dailyReturnReinv']
#数据取交集，匹配x和y，找到交集
data=data_timex.join(data_timey, how='inner')   
data.dropna(axis=0,inplace=True,how='any')
j=1
buy=[]
time=list(data.index)#这就是我们所选取的时间
#第二步：策略
strategy_return=[]
benchmark_strategy_return=[]
for i in range(len(time)-2): 
    data_x=DataAPI.MktStockFactorsOneDayGet(tradeDate=time[i],secID=u"",ticker=u"",
field=u"",pandas="1")#提取数据x
    data_x=data_x.drop(['ticker','tradeDate'],axis=1)#删除非数值指标
    data_x.set_index('secID',inplace=True)#设股票代码索引
    data_y=DataAPI.EquRetudGet(secID=u"",ticker=u"",listStatusCD=u"",
beginDate=time[i+1],endDate=time[i+1],dailyReturnNoReinvLower=u"",dailyReturnNoReinvUpper=u"",dailyReturnReinvLower=u"",dailyReturnReinvUpper=u"",isChgPctl=u"",field=u"",pandas="1")#提取数据y
    data_y.set_index('secID',inplace=True)
    data_y=data_y['dailyReturnReinv']
    data_y.dropna(axis=0,inplace=True)
    #数据取交集，匹配x和y，找到交集
    data=data_x.join(data_y, how='inner') 
    data.dropna(axis=1,inplace=True,how='all')
    data.dropna(axis=0,inplace=True,how='any')#hang
    data_y=data['dailyReturnReinv']
    data_x=data.drop(['dailyReturnReinv'],axis=1)
    #我们用今天的指标跟明天的收益率做预测
    data_x=pd.DataFrame(data_x)
    data_xx=preprocessing.scale(data_x.values,axis=0)#数据标准化
    data_x=pd.DataFrame(data_xx,index=data_x.index,columns=data_x.columns)
    data_y[data_y<0]=-1#因变量指标
    data_y[data_y>=0]=1
    #我们用昨天的数据预测今天的进行训练，然后用今天的数据预测明天的收益率
    data_yan=DataAPI.MktStockFactorsOneDayGet(tradeDate=time[i+1],secID=u"",ticker=u"",
field=u"",pandas="1")#预测值
    data_yan=data_yan.drop(['ticker','tradeDate'],axis=1)
    data_yan.set_index('secID',inplace=True)
    data_yan.dropna(axis=1,inplace=True,how='all')#列删除
    data_yan.dropna(axis=0,inplace=True,how='any')#hang
    data_yan=pd.DataFrame(data_yan) 
    #要保证训练指标和预测指标一致，于是做如下处理： 
    # 由于训练和预测的特征数量可能不同，所以需要columns的并集对列标签进行调整
    character_union = data_x.columns.union(data_yan.columns)
    data_x= data_x.reindex(columns=character_union, fill_value=0)
    data_yan = data_yan.reindex(columns=character_union, fill_value=0) 
    data_yany=preprocessing.scale(data_yan.values,axis=0)#数据标准化
    data_yan=pd.DataFrame(data_yany,index=data_yan.index,columns=data_yan.columns)#把标准化的数据加上索引和指标

    if abs(data_y.sum())==len(data_y.values):#如果都为-1或者1，则无法训练因此我们取当天股票收益率的平均值为今天收益率
        r=0
        strategy_return.append(r)
        #print("全涨或者跌")
    else:
        clf=RandomForestClassifier()
        # 训练模型
        #嵌入指标筛选
        feat_labels=data_x.columns
        clf= clf.fit(data_x, data_y)
        importance=clf.feature_importances_  
       # model = SelectFromModel(clf_tree, prefit=True)
       # X_new = model.transform(data_x)
        #X_new.shape               
        indices=np.argsort(importance)[::-1]
        feature=[]
        for f in range(data_x.shape[1]): 
            feature.append(feat_labels[indices[f]])
        feature=feature[:40]#选取的指标的个数
        data_x=pd.DataFrame(data_x[feature])
        data_yan=pd.DataFrame(data_yan[feature])
        #上为指标筛选
        clf.fit(data_x.values, data_y.values.astype(int))
        prediction = clf.predict_proba(data_yan.values)
        # print(pd.DataFrame(prediction).iloc[:,1])
        data_yan['prediction']=pd.DataFrame(prediction).iloc[:,1].values
        data_yan.sort_values(by='prediction',inplace=True,ascending=False)
        buylist=data_yan[data_yan['prediction']>0.995].index
        buy.append(buylist)#(空出第一列)
        data_change=DataAPI.EquRetudGet(secID=buylist,ticker=u"",listStatusCD=u"",
beginDate=time[i+2],endDate=time[i+2],dailyReturnNoReinvLower=u"",dailyReturnNoReinvUpper=u"",dailyReturnReinvLower=u"",dailyReturnReinvUpper=u"",isChgPctl=u"",field=u"",pandas="1")#我们收益率用推迟一天的，可以直接匹配
        data_change.set_index('secID',inplace=True)#接下来进行数据合并，取公共有数据的股票
        data_change=data_change['dailyReturnReinv']
        data_change.fillna(0,inplace=True)#hang  .dropna(,axis=,how='any',inplace=True)
        #print(data_change)
        length=len(buylist)
        #print(data_change)
        if length==0:#如果今天买入股票为零，那么我们利润为0
            r=0
            strategy_return.append(r)
            j=j+1
        else:
            danrishouyi=[]
            if i==0:
                r= data_change.values.sum()/(length)
                danrishouyi.append(r)
            else:
                e=0
                for h in buylist:
                    e=e+1
                    if h in buy[i-1]:#因为第一列为空，i+1的前一列为i
                        r=data_change[e-1]#e从1开始，而data_change从0开始
                        danrishouyi.append(r)
                    else:
                        r=data_change[e-1]  #-0.004
                        danrishouyi.append(r)
            shouyi=sum(danrishouyi)/(length)# length+1
            strategy_return.append(shouyi)
#计算总收益率
all_return=(pd.DataFrame(strategy_return)+1).cumprod()
print(j) 
print('不买股票的次数')
```

```python
#大盘平均收益率
benchmark_strategy_return=[]
for i in range(len(time)):
    benchmark_change=DataAPI.EquRetudGet(secID=u"",ticker=u"",listStatusCD=u"",
beginDate=time[i],endDate=time[i],dailyReturnNoReinvLower=u"",dailyReturnNoReinvUpper=u"",dailyReturnReinvLower=u"",dailyReturnReinvUpper=u"",isChgPctl=u"",field=u"",pandas="1")#我们收益率用推迟一天的，可以直接匹配
    benchmark_change.set_index('secID',inplace=True)#接下来进行数据合并，取公共有数据的股票
    benchmark_change=benchmark_change['dailyReturnReinv']
    benchmark_change.dropna(axis=0,inplace=True,how='any')#hang
    benchmark_length=len(benchmark_change)
    benchmark_r= benchmark_change.values.sum()/(benchmark_length)
    benchmark_strategy_return.append(benchmark_r)

benchmark_strategy_return=pd.DataFrame(benchmark_strategy_return)+1
benchmark_all_return=benchmark_strategy_return.cumprod()
#print(benchmark_all_return)
```
