# 价差偏离度因子

前几天逛社区看到了有关价差偏离这个因子的帖子 作者参考了《研究报告：东方证券-因子选股系列研究之七：投机、交易行为与股票收益（下）》-160512 这篇研报 
有关因子数据的计算方式可以参考社区里的帖子 https://uqer.datayes.com/community/share/574c193b228e5b86a71fd930
另外一个帖子里 https://uqer.datayes.com/community/share/57d54eb4228e5b0494fb85db 
作者指出了上一个帖子里面存在的数据计算的问题 以及回测结果 其中没有解决动态因子池的问题以及IC 收益分组 以及市值暴露 和中性化的测试  
下面稍微简述一下因子的计算过程：
首先介绍一下该因子的含义：
	价差偏离度(SpreadBias): 借鉴统计套利的思想 目的在于找到某只股票相对于[同类型]的股票的相对高估低估的程度 。具体可参考研报
    第一步：我们将在每个月末对每只股票在全市场上来匹配出同类型的股票，这(10)只股票加在一起组成了这支股票的标的。标的值就是股价的平均值  这样每只股票都拥有了自己的标的值
    第二步：获取下一个整月的所有股票的 标的价格  和 这支股票的价格 计算 PriceSpread=ln(StockPrice）−ln(ReferencePrice)
这里面的逻辑是：在下一个整月认为组成标的股票不改变，这样每天都会有跟踪价差
经过上面两个步骤就可以得到一个dataframe 记录这每个交易日的 所有股票的跟踪价差  下面我们将在每一个月末计算真正的因子数据。
	第三步：SpreadBias = (PriceSpread−mean(PriceSpread))/std(PriceSpread)  这个窗口选(60)日   经过第三步之后就获得了月度的因子数据 接下来就要进行其因子分布  IC return 市值分布 等测试 期间会对因子进行相应的去极值 中性化 标准化 等处理
    补充：大家看到这里会发现 如何定义上面所谓的同类型的算法没有交代，这里将先介绍研报中的定义方式，然后给出自己的一些想法。
研报定义方式：两个股票间距离=1-两个股票在过去(250)个交易日收益率的pearson相关系数。选择距离最小的10只个股作为S的标的组合。
一些看法：首先股票的涨跌不是各自独立的，与板块概念有关并且是轮动的，那么很明显某板块滞胀的就会有补涨的需求，反之。那么当某只股票的近期表现弱于其相似的股票那么短期内就会补涨。  相似的股票有相似的历史表现（研报中以涨跌幅来定义相似，这里举个例子：比如一些成交量很小的横盘调整很久的股票一旦赶上风口起飞的时候，这种匹配方式将存在很严重的滞后性失去很多的机会，总的来说就是这种匹配方式不太能够很好的找到背后的板块概念的效应） ，逻辑应该是当下这个月那个板块或者概念比较好但是某个相关的股票表现并不好于是认为它会补涨。 
    (注意：计算的时候尽量限制api接口的调用因为调用会耗时并且对调用次数会有限制 最好是把数据一次性都拿出来然后计算)

参数：250 10 60 股票池:交易HS300 在A里面找匹配   一共五个参数   下面的分析选用 50 10 30 ZZ500 ZZ500（没办法就这样都跑了十二个小时 爆炸）

```python
#这个是获取数据的函数 大家可以存到lib里方便以后调用
def getUqerStockFactors(begin, end, factor, universe=None):
    # 拿取上海证券交易所日历
    cal_dates = DataAPI.TradeCalGet(exchangeCD=u"XSHG", beginDate=begin, endDate=end)
    cal_dates = cal_dates[cal_dates['isOpen']==1].sort('calendarDate')
    # 工作日列表
    cal_dates = cal_dates['calendarDate'].values.tolist()                             
    print  factor + ' will be calculated for ' + str(len(cal_dates)) + ' days:'
    count = 0
    secs_time = 0
    start_time = time.time()
    
    # 按天拿取因子数据，并保存为一个dataframe
    df = pd.DataFrame()
    for dt in cal_dates:
        # 拿取数据dataapi，必要时可以使用专业版api
        dt_df = DataAPI.MktEqudGet(tradeDate=dt, secID='', 
                                                 field=['tradeDate', 'secID']+[factor])
        if df.empty:
            df = dt_df
        else:
            df = df.append(dt_df)
        
        # 打印进度部分，每200天打印一次
        if count > 0 and count % 200 == 0:
            finish_time = time.time()
            print count,
            print '  ' + str(np.round((finish_time-start_time) - secs_time, 0)) + ' seconds elapsed.'
            secs_time = (finish_time-start_time)
        count += 1
        
    # 提取所需的universe对应的因子数据
    df = df.set_index(['tradeDate','secID'])[factor].unstack()
    if universe:
        universe = list(set(universe) & set(df.columns))
        df = df[universe]
    df.index = pd.to_datetime(df.index, format='%Y-%m-%d')
    
    # # 将上市不满三个月的股票的因子设置为NaN
    equ_info = DataAPI.EquGet(equTypeCD=u"A",secID=u"",ticker=u"",listStatusCD=u"",field=u"",pandas="1")
    equ_info = equ_info[['secID', 'listDate', 'delistDate']].set_index('secID')
    equ_info['delistDate'] = [x if type(x)==str else end for x in equ_info['delistDate']]
    equ_info['listDate'] = pd.to_datetime(equ_info['listDate'], format='%Y-%m-%d')
    equ_info['delistDate'] = pd.to_datetime(equ_info['delistDate'], format='%Y-%m-%d')
    equ_info['listDate'] = [x + timedelta(90) for x in equ_info['listDate']]
    
    for sec in df.columns:
        if sec[0] not in '036':
            continue
        sec_info = equ_info.ix[sec]
        df.loc[:sec_info['listDate'], sec] = np.NaN
        if end > sec_info['delistDate'].strftime('%Y%m%d'):
            df.loc[sec_info['delistDate']:, sec] = np.NaN
    return df
```

```python
#下面是计算价差的函数
cal_dates = DataAPI.TradeCalGet(exchangeCD=u"XSHG", beginDate='2014-01-01', endDate='2017-07-01')
cal_dates_day=cal_dates[cal_dates['isOpen']==1]['calendarDate'].values.tolist()
cal_dates_mon = cal_dates[cal_dates['isMonthEnd']==1]
cal_dates_mon =cal_dates_mon['calendarDate'].values.tolist()
def distance(stk1,stk2,Date,peri=50):#给两个股票代码 这一天的日期 和窗口长度   1-50日的两个股票的涨跌幅的pearson相关系数  作为距离
    if stk1==stk2:
        return 0.0
    else:
        cal = Calendar('China.SSE')
        trueperiod=Period('-'+str(peri)+'B')
        startdate=cal.advanceDate(Date,trueperiod)
        startdate =startdate.toDateTime()
        ss=startdate.strftime('%Y-%m-%d')
        #下面直接在获得的大的每日涨幅的表中调用 不通过api获取 这样避免了调用次数的限制
        if ss in daily_return.index:
            for j in range(0,len(daily_return.index)):
                #print daily_return.index[j]
                if ss==daily_return.index[j]:
                    temp=j
                    
            date=daily_return.index[temp:temp+peri]
            #print date
            df1=pd.DataFrame(daily_return.loc[date,stk1])
            #print date
            df2=pd.DataFrame(daily_return.loc[date,stk2])
            df1=df1.merge(df2,left_index=True,right_index=True)
            df1.dropna(inplace=True)
            s=df1[stk1].corr(df1[stk2],method='pearson')
            dis=1.0-s
            return dis
def targetlist(stk,Date,N=10):#给股票和日期 将在整个股票池里面选出 前N位的匹配股票
    universe=set_universe('ZZ500',Date)
    target={}
    for ticker in universe:
        diss=distance(stk,ticker,Date)
        target[ticker]=diss
    #print target
    dic= sorted(target.iteritems(), key=lambda d:d[1])    #因为字典不支持排序，我们先把字典转换成二元数组进行排序
    dics=[item[0] for item in dic ]
    return (dics[0:N] if stk not in dics else dics[1:N+1])
def reference_price(target,Date): #传入一个list某几只股票 计算参考价格 注意这里面Date是一个list（包含着下个月的全部日期）这样返回的就是下个月的所有天的参考价格
    #print target
    return np.log(close_ZZ500.loc[Date,target].T.mean())
    

price_spread=pd.DataFrame(index=cal_dates_day,columns=ZZ500_list)
#下面的目的在于把 price_spread 填满
for i in range(0,len(cal_dates_mon)-1):#是月末的日期  我要获得下一个月的所有日期
    date_in_month = DataAPI.TradeCalGet(exchangeCD=u"XSHG", beginDate=cal_dates_mon[i], endDate=cal_dates_mon[i+1])
    date_in_month=date_in_month[date_in_month['isOpen']==1]['calendarDate'].values.tolist()[1:]
    #此时 date_in_month 是下个月的全部交易日日期    cal_dates_mon[i]是此时的月末日期    
    universe=set_universe('ZZ500',cal_dates_mon[i])
    #targetlist(stk,date)  reference_price(target,Date)
    for stk in universe:
        ref_list=targetlist(stk,cal_dates_mon[i])
       
        re_p=reference_price(ref_list,date_in_month)#series 格式
        #print close_ZZ500.loc[date_in_month,stk]
        s_p=np.log(close_ZZ500.loc[date_in_month,stk])
        price_spread.loc[date_in_month,stk]=s_p-re_p
```

下面进行因子的分析 
IC数据 显示还是存在着预测性的 不过毕竟三年的数据的说服力不是特别的大

```python
begin='2013-11-11'
end='2017-02-01'
cal_dates = DataAPI.TradeCalGet(exchangeCD=u"XSHG", beginDate=begin, endDate=end)
cal_dates = cal_dates[cal_dates['isOpen']==1].sort('calendarDate')
cal_dates = cal_dates[cal_dates['isMonthEnd']==1]
cal_dates = cal_dates['calendarDate'].values.tolist()
universe=ZZ500_list
closePrice = DataAPI.MktEqudAdjGet(secID=universe, isOpen="1", beginDate=begin, endDate=end, field=u"tradeDate,secID,closePrice",pandas="1")
closePrice = closePrice.pivot(index='tradeDate', columns='secID', values='closePrice')
forward_20d_return_data = closePrice.shift(-20)/closePrice-1
forward_20d_return_data = forward_20d_return_data.ix[cal_dates, :]
forward_20d_return_data.index = pd.to_datetime((forward_20d_return_data.index))
```

```python

ic_data = pd.DataFrame(index=forward_20d_return_data.index, columns=['IC','pValue'])

# 计算相关系数
for dt in ic_data.index:
    dt=dt.strftime('%Y-%m-%d')
    if dt in factor.index:
        tmp_factor = factor_final.ix[dt]
        tmp_ret = forward_20d_return_data.ix[dt]
        cor = pd.DataFrame(tmp_factor)
        ret = pd.DataFrame(tmp_ret)
        cor.columns = ['corr']
        ret.columns = ['ret']
        cor['ret'] = ret['ret']
        #cor = cor[~np.isnan(cor['corr'])][~np.isnan(cor['ret'])]
        if len(cor) < 5:
            continue

        ic, p_value = st.spearmanr(cor['corr'],cor['ret'])   # 计算秩相关系数 RankIC
        ic_data['IC'][dt] = ic
        ic_data['pValue'][dt] = p_value
    
print 'mean of IC: ', ic_data['IC'].mean(), '；',
print 'median of IC: ', ic_data['IC'].median()
print 'the number of IC(all, plus, minus): ', (len(ic_data), len(ic_data[ic_data.IC>0]), len(ic_data[ic_data.IC<0]))


# 每一天的**因子**和**之后20日收益**的秩相关系数作图

fig = plt.figure(figsize=(16, 6))
ax1 = fig.add_subplot(111)

lns1 = ax1.plot(ic_data[ic_data>0].index, ic_data[ic_data>0].IC, '.r', label='IC(plus)')
lns2 = ax1.plot(ic_data[ic_data<0].index, ic_data[ic_data<0].IC, '.b', label='IC(minus)')

lns = lns1+lns2
labs = [l.get_label() for l in lns]
ax1.legend(lns, labs, bbox_to_anchor=[0.6, 0.1], loc='', ncol=2, mode="", borderaxespad=0., fontsize=12)
ax1.set_ylabel(u'相关系数', fontproperties=font, fontsize=16)
ax1.set_xlabel(u'日期', fontproperties=font, fontsize=16)
ax1.set_title(u"全A股票池CGO因子的截面月度IC", fontproperties=font, fontsize=16)
ax1.grid()
```

下面是分组的收益，可以看出其实结果并不是很单调，可见这个因子相对于社区里的其他因子效果表现不好 

```python
n_quantile = 10
# 统计十分位数
cols_mean = [i+1 for i in range(n_quantile)]
cols = cols_mean

excess_returns_means = pd.DataFrame(index=factor.index, columns=cols)

# 计算因子分组的超额收益平均值
for dt in excess_returns_means.index:
    qt_mean_results = []
    if dt in factor_final.index and dt in forward_20d_return_data.index:
    # ILLIQ去掉nan
        tmp_factor = factor_final.ix[dt].dropna()

        tmp_return = forward_20d_return_data.ix[dt].dropna()
        tmp_return = tmp_return[tmp_return<0.6]
        tmp_return_mean = tmp_return.mean()

        pct_quantiles = 1.0/n_quantile
        for i in range(n_quantile):
            down = tmp_factor.quantile(pct_quantiles*i)
            up = tmp_factor.quantile(pct_quantiles*(i+1))
            i_quantile_index = tmp_factor[(tmp_factor<=up) & (tmp_factor>=down)].index
            mean_tmp = tmp_return[i_quantile_index].mean() - tmp_return_mean
            qt_mean_results.append(mean_tmp)

        excess_returns_means.ix[dt] = qt_mean_results

excess_returns_means.dropna(inplace=True)

# 因子分组的超额收益作图
fig = plt.figure(figsize=(12, 6))
ax1 = fig.add_subplot(111)

excess_returns_means_dist = excess_returns_means.mean()
excess_dist_plus = excess_returns_means_dist[excess_returns_means_dist>0]
excess_dist_minus = excess_returns_means_dist[excess_returns_means_dist<0]
lns2 = ax1.bar(excess_dist_plus.index, excess_dist_plus.values, align='center', color='r', width=0.35)
lns3 = ax1.bar(excess_dist_minus.index, excess_dist_minus.values, align='center', color='g', width=0.35)

ax1.set_xlim(left=0.5, right=len(excess_returns_means_dist)+0.5)
# ax1.set_ylim(-0.008, 0.008)
ax1.set_ylabel(u'超额收益', fontproperties=font, fontsize=16)
ax1.set_xlabel(u'十分位分组', fontproperties=font, fontsize=16)
ax1.set_xticks(excess_returns_means_dist.index)
ax1.set_xticklabels([int(x) for x in ax1.get_xticks()], fontproperties=font, fontsize=14)
ax1.set_yticklabels([str(x*100)+'0%' for x in ax1.get_yticks()], fontproperties=font, fontsize=14)
ax1.set_title(u"因子超额收益", fontproperties=font, fontsize=16)
ax1.grid()
```
