## 先简单介绍一下scrapy
- scrapy是一个爬虫框架，通俗说就是写爬虫的半成品，在上面稍微改改就能爬了，里面很多东西都带你写好了，不用怎么修改就ok，听起来是不是很方便？
### scrapy有下面几个优势
- 它让我们少写很多代码，它是基于twist框架做的，是一种单进程异步多线程的请求方式，如果手写这种单进程异步多线程会非常恶心，我上一篇就没写，手写要用到ansio那些乱七八糟的 代码前面要每句要加上ansy什么玩意的看着就让人恶心，看上去完全不像正常的代码。
- 它把处理方式都归类了，爬虫的写爬虫部分，中间件写中间件，item处理就处理，设置里面有很多非常好用的设置，条理清晰让人舒服。
- 它有更多牛b的扩展，比如scrapy-redis可以把里面调度器变成redis更好的管理需要爬取的url。还有scrapyd可以分布式把爬虫任务部署到多台机器上进行爬取各个网站。

## 1.在items.py下按照例子输入要爬取的东西

```python
import scrapy


class PachongtestItem(scrapy.Item):
    # define the fields for your item here like:
    # name = scrapy.Field()
    name = scrapy.Field()
    shijian = scrapy.Field()
    kfincome = scrapy.Field()
    pass
```

## 2.在spiders下的fif.py参考以下代码 就是把要爬的网址yield交给parse 然后parse会自己判断如果是Item就给pipeline 如果是request就看你callback给谁，使用meta可以传递参数 非常简单

```python
import scrapy
from scrapy.http import Request
from ..items import PachongtestItem

class FifSpider(scrapy.Spider):
    name = 'fif'
    allowed_domains = ['quotes.money.163.com']
    # start_urls = ['http://quotes.money.163.com/']
    # 时间
    #response.css('#scrollTable > div.col_r > table  > tr:nth-child(1)>th::text').extract()
    # 扣非
    #response.css('#scrollTable > div.col_r > table  > tr:nth-child(12) > td::text').extract()
    ti = []
    secdata = open('C:\文件路径\sec.txt', 'r', encoding='utf-8')
    for h in secdata.readlines():
        h.encode('utf-8').decode('utf-8-sig')
        h = h.replace('.XSHG', '')
        h = h.replace('.XSHE', '')
        h = h.replace('\n', '')
        ti.append(h)
    if ti[-1] == "":
        ti.pop(-1)
    def start_requests(self):
        for i in self.ti:
            url = 'http://quotes.money.163.com/f10/zycwzb_%s.html#01c01' % (i)
            yield  Request(url,callback=self.parse,meta={'name':i})
    def parse(self, response):
        item = PachongtestItem()
        item['name'] = response.meta['name']
        item['shijian'] = response.css('#scrollTable > div.col_r > table  > tr:nth-child(1)>th::text').extract()
        item['kfincome'] = response.css('#scrollTable > div.col_r > table  > tr:nth-child(12) > td::text').extract()
        return item
```

## 3.1 在pipelines.py下面处理入库操作。跟上一篇区别不大，就是写到类里面加个self而已

```python
import  pymysql
class PachongtestPipeline(object):
    def __init__(self):
        self.db = pymysql.connect("localhost", "root", "123456", "test2")
        self.cursor = self.db.cursor()
    def process_item(self, item, spider):
        self.cursor.execute("DROP TABLE IF EXISTS `%s`"%(item['name']))
        sql = """CREATE TABLE `%s` (DATETIME  CHAR(20) ,KFINCOME CHAR(20) )"""%(item['name'])
        self.cursor.execute(sql)
        zii = zip(item['shijian'],item['kfincome'])
        for i in zii:
            sql = """INSERT INTO `%s`(DATETIME,KFINCOME)VALUES ('%s','%s')""" % (item['name'], i[0], i[1])
            self.cursor.execute(sql)
            self.db.commit()
```

# 五、数据导入优矿
- mysql的导入方法跟上一篇一样，这里介绍mongodb的导入方法

```python
import pandas  as pd
import  numpy as np
from pymongo import MongoClient

con = MongoClient()
db = con.ffkf

collection = list(db.collection_names())
df = pd.DataFrame()
secli = []
for i in collection:
    res = list(db[i].find({},{"_id":0}))
    secli.append('sec'+i)
    tmp1 = list(res[0].keys())
    tmp2 = list(res[0].values())
    df1 = pd.DataFrame(tmp2)
    df1 = df1.T
    df1.columns = tmp1
    df = pd.concat((df, df1), axis=0)
secli = np.array(secli)
df = df.set_index(secli)
df.to_csv('final2.csv')
```

# 六、一些TIPS
- settings里面最后面注释的几行http cache那些可以开启能增加爬取速度
- settings里面并发请求数可以自己调一下试试玩也能增加速度
- settings里面有个robots协议，如果网站开启协议你就会爬不到，把遵循robot协议关了
- 觉得在cmd里运行不方便？有几种方法
