```python
import datetime
import matplotlib.pyplot as plt
import numpy as np
from CAL.PyCAL import *
import matplotlib as mpl
import pandas as pd
```

```python
manager = DataAPI.FundManagerGet(secID=u"163402.XSHE",ticker=u"",field=u"",pandas="1")
manager_brief = manager[['name','accessionDate','dimissionDate']]
manager_brief['dimissionDate'].fillna(str(datetime.date.today()),inplace=True)
manager_brief['dimissionDate'] = pd.to_datetime(manager_brief['dimissionDate']).astype(datetime.datetime)
manager_brief['accessionDate'] = pd.to_datetime(manager_brief['accessionDate']).astype(datetime.datetime)

fig,(ax1,ax2) = plt.subplots(2,1,sharex=True,figsize=(24,12))
ypos = np.arange(len(manager_brief.index))
ax2.set_yticks(ypos)
ax2.set_yticklabels(manager_brief['name'].apply(lambda x: x.decode('utf-8')), fontproperties = font)
x_zero = mpl.dates.date2num(manager_brief['accessionDate'].min())
start = mpl.dates.date2num(manager_brief['accessionDate'])
width= mpl.dates.date2num(manager_brief['dimissionDate']) - mpl.dates.date2num(manager_brief['accessionDate'])

ax2.barh(ypos,width,left= start)
yeb = DataAPI.FundNavGet(dataDate=u"",secID=u"163402.XSHE",ticker=u"",beginDate=u"20050101",endDate=u"20171007",field=u"",pandas="1")
# yeb = DataAPI.FundDivmGet(dataDate=u"",secID=u"163402.XSHE",ticker=u"",beginDate=u"20050101",endDate=u"20171007",field=u"",pandas="1")
yeb['endDate'] = pd.to_datetime(yeb['endDate']).astype(datetime.datetime)
ax1.plot(yeb['endDate'],yeb['ACCUM_NAV'])
# yeb[['endDate','dailyProfit']].plot('endDate','dailyProfit',figsize=(12,6),ax=ax1)
ax2.xaxis_date()
# format date to ticks on every 7 days

rule = mpl.dates.rrulewrapper(mpl.dates.MONTHLY, interval=3)
loc = mpl.dates.RRuleLocator(rule)
formatter = mpl.dates.DateFormatter("%Y-%m-%d")
ax2.xaxis.set_major_locator(loc)
ax2.xaxis.set_major_formatter(formatter)
xlabels = ax2.get_xticklabels()
plt.setp(xlabels, rotation=30, fontsize=9)
plt.show()
```
