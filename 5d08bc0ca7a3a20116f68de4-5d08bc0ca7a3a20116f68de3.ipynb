# 行业盈利能力计算：截面与历史平均

```python
###各行业费用占比
### xscbl=销售成本占比，ssfyl=销售费用率，glfyl=管理费用率，cwfyl=财务费用率
import pandas as pd
sec=set_universe('A')
df1=DataAPI.FdmtISGet(secID=sec,reportType=u"A",endDate=u"",beginDate=u"",publishDateEnd=u"",publishDateBegin=u"",endDateRep="",beginDateRep="",beginYear="2018",endYear="2018",fiscalPeriod="",field=u"endDate,ticker,tRevenue,COGS,sellExp,adminExp,finanExp",pandas="1")
df2=DataAPI.EquIndustryGet(secID=sec,industryVersionCD=u"010303",industry=u"",industryID=u"",industryID1=u"",industryID2=u"",industryID3=u"",intoDate=u"20180101",field="endDate,ticker,industryName1",pandas="1")
df=df1.merge(df2,left_on=['ticker'], right_on=['ticker'], how='inner')
df['xscbl']=df['COGS']/df['tRevenue']
df['ssfyl']=df['sellExp']/df['tRevenue']
df['glfyl']=df['adminExp']/df['tRevenue']
df['cwfyl']=df['finanExp']/df['tRevenue']
df[['ticker','xscbl','ssfyl','glfyl','cwfyl','industryName1']]
fyjg=pd.DataFrame(index=df['industryName1'].drop_duplicates(),columns=[['xscbl','ssfyl','glfyl','cwfyl']])
for x in df['industryName1'].drop_duplicates():
    fyjg.loc[x]=df[df['industryName1']==x][['xscbl','ssfyl','glfyl','cwfyl']].mean()
for x in ['xscbl','ssfyl','glfyl','cwfyl']:
     print fyjg[x].argmax()
```
