# 上市公司职工薪酬排行版

```python
import pandas as pd

# 从接口获取员工总数表
df1 = DataAPI.EquShareEmployeesGet(secID=u"",beginDate=u"20170101",endDate=u"20190101",field=u"",pandas="1")
df1 = df1.loc[df1["typeValue"] == "员工总数"]

#从接口获取2018第三季度财务报告中的应付职工薪酬表
df2 = DataAPI.FdmtBSGet(secID=u"",reportType=u"",endDate=u"20190101",beginDate=u"20180901",publishDateEnd=u"",publishDateBegin=u"",endDateRep="",beginDateRep="",beginYear="",endYear="",fiscalPeriod="",field=u"secID,secShortName,publishDate,payrollPayable",pandas="1")

#合并员工总数表及应付职工薪酬表
df3 = pd.merge(df2, df1, how='right', on=['secShortName'])
df3.head(20)

#获取上市公司对应的行业
df4 = DataAPI.EquIndustryGet(secID=u"",industryVersionCD=u"010303",industry=u"",industryID=u"",industryID1=u"",industryID2=u"",industryID3=u"",intoDate=u"",field=u"",pandas="1")
df4 = df4[['secID','secShortName','industryName1']]

#合并以上所有数据表
df5 = pd.merge(df4,df3,how='right',on = ['secShortName'])

#增加一列计算列，平均月薪酬支出，计算逻辑为依照第三季度财报，职工薪酬支出除以9，调整单位为万，除以10000，再除以员工总数，得出平均月薪酬支出salary
df5["salary"] = df5.payrollPayable /90000 / df5.personNum

# 数据去重
df5 = df5.drop_duplicates(subset=['secID_x'],keep='first')
# 取数,排序，月薪酬支出最高的前30家公司
df5 = df5[['secID','secShortName','industryName1','personNum','salary']]
df5.sort_values(by="salary",ascending = False).head(70)
```
