**Alpha#41**
$$ \large Alpha41: \sqrt{high * low} - vwap$$



公式很简单，用前一天的 high和low的**几何平均值**，减去**成交量加权平均价vwap**

1.首先解释下vwap：

成交量加权平均价是将多笔交易的价格按各自的成交量加权而算出的平均价，若是计算某一证券在某交易日的VWAP，将当日成交总值除以总成交量即可。VWAP可作为交易定价的一种方法，亦可作为衡量机构投资者或交易商的交易表现的尺度。英文亦称为dynamic time and sales。采用日内交易量加权平均价格能够更加有效真实的反应股票的价格运动。

假如我们把成交量看作“投票”，我们为自己心中的目标价位投票。显然，成交量高的价格，是“票数”最高的价格，可以看作是**更接近市场整体的估计价位。**

这式子中，可以把vwap作为参考，我们通过计算一个“价格”，**以它偏离vwap的程度**为权重建立头寸。

2.比较难理解的是high和low的几何平均值，不妨来比较下算术平均值。我们都知道有$$ \large \frac{a+b}{2} \geqslant \sqrt{ab} $$

可以看出，几何平均值比起算术平均值，是偏“保守”的。

那么为什么不用收盘价，不用high、low的算术平均，而用high、low的几何平均呢？我们不妨来观察下他们的区别。

```python
from __future__ import division
import numpy as np
import matplotlib.pyplot as plt

data = DataAPI.MktEqudGet(secID=u"",ticker=u"000001",beginDate=u"20120101",endDate=u"20151220",isOpen="",field=u"",pandas="1")  
data = data.set_index('tradeDate')
data['Mean'] = (data['highestPrice'] + data['lowestPrice'])/2
data['GeometricMean'] = np.sqrt(data['highestPrice']*data['lowestPrice'] )

data['Mean'] = data['Mean'].replace(0,np.nan).dropna()  # 去除停牌
data['GeometricMean'] = data['GeometricMean'].replace(0,np.nan).dropna()

#data['closePrice'].plot(figsize=(22,10))
data['Mean'].plot(figsize=(22,10))
data['GeometricMean'].plot()
plt.legend(loc=0)
```

好吧，两种平均的方法差距很小~ 我们再看看几何平均和直接用收盘价的区别：

```python
data['closePrice'].plot(figsize=(22,10))
data['GeometricMean'].plot()
plt.legend(loc=0)
```

细心的同学会发现，**在极值处，尤其是到达极值后立马出现较大反转的地方，high和low的几何平均往往表现出：在极大值处比收盘价小；在极小值处比收盘价大的现象！**

最后补充下前复权的概念。

**前复权：**

复权就是对股价和成交量进行权息修复，按照股票的实际涨跌绘制股价走势图,并把成交量调整为相同的股本口径。股票除权、除息之后，股价随之产生了变化，但实际成本并没有变化。如：原来20元的股票，十送十之后为10元，但实际还是相当于20元。从K线图上看这个价位看似很低，但很可能就是一个历史高位。

前复权，就是持现有价位不变，将以前的价格缩减，将除权前的K线向下平移，使图形吻合，保持股价走势的连续性。
