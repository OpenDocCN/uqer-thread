```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import datetime as dt
import CAL.PyCAL as C_cal # CAL provides classes and functions dealing with date, time, and calendars. 

sns.set()
pd.set_option('display.width', 200)
cal = C_cal.Calendar('China.SSE')
```

```python
start_date = '20180801' # 开始日期
end_date = (dt.datetime.today() - dt.timedelta(days=1)).strftime('%Y%m%d') # 截止日期

# 交易所交易日历 
"""
XSHG，XSHE，CCFX，XDCE，XSGE，XZCE，XHKG。XSHG表示上海证券交易所，XSHE表示深圳证券交易所，CCFX表示中国金融期货交易所，XDCE表示大连商品交易所，XSGE表示上海期货交易所，XZCE表示郑州商品交易所，XHKG表示香港证券交易所。可同时输入多个证券交易所,可以是列表
"""
TradeCal=DataAPI.TradeCalGet(exchangeCD="XSHG",beginDate=start_date,endDate=end_date,field="calendarDate,isOpen",pandas="1")
TradeCal = TradeCal.loc[TradeCal['isOpen'] == 1,:] # TradeCal = TradeCal[TradeCal['isOpen'] == 1]

Dates = map(lambda x: x[0:4]+x[5:7]+x[8:10], TradeCal['calendarDate'])
PB_PE = pd.DataFrame(index=Dates, data=0, columns=['total PB(LFLO)','total PE(LFLO)','shangzheng','shenzheng','chuangye','hs300','shzh50','zhzh500'])
for date in Dates:
    """
    上证50(SH50)、上证180(SH180)、沪深300(HS300)、中证500(ZZ500)、创业板(CYB)、中小板(ZXB)、全A股(A)，如"HS300"；另外，支持自定义     指数代码，可以获得date日期的成分股列表，如"000001.ZICN"；对于预设     行业实例（IndSW/IndZJH/IndZZ）及预设指数实例（IdxCN），可直接使     用date (datetime.datetime): 交易日
    """
    universe = set_universe('A', date) # + set_universe('ZZ500', date)
    
    # 获取多只股票历史上某一天的因子数据
    data = DataAPI.MktStockFactorsOneDayGet(tradeDate=date,secID=universe[:],field=['ticker','LFLO','PB','PE'],pandas="1").set_index('ticker')
    """
    LCAP	float	对数市值（Natural logarithm of total market values）。计算方法：对数市值=市值的对数。属于价值类因子。数据更新     时间：T日21点30分。
    LFLO	float	对数流通市值（Natural logarithm of float market values）。计算方法：对数流通市值=流通市值的对数。属于价值类因     子。数据更新时间：T日21点30分。
    """
    
    data_total = data.shape[0]
    data.dropna(inplace=True)
    # 数据量不足80%时，不具有代表性
    if float(data.shape[0]) / data_total < 0.8:
        print('数据量不足80%. \n')
        continue
        
     # 细节处理：负值极值处理
    data['PB'][data['PB'] < 0] = 0
    data['PE'][data['PE'] < 0] = 0
    pb = winsorize(data['PB'].to_dict()) # 返回 dic
    pe = winsorize(data['PE'].to_dict())
    data['PB'][pb.keys()] = pb.values() # data['PB'] 是 类似DataFrame，即 有索引的列(可理解为原有DataFrame的为删除索引的子集)； ？？！！这是什么类型？？
    data['PE'][pe.keys()] = pe.values()
    
    # 权重
    data['LFLO'] = data['LFLO'] / sum(data['LFLO'])
    
    # 获取指数日线行情信息，包含昨收价、开盘价、最高价、最低价、收盘价、成交量、成交金额等字段。
    MktIdxdGet_1=DataAPI.MktIdxdGet(ticker=['000001','399001','399006','000300','000016','000905'],tradeDate=date,field='closeIndex',pandas="1")
    """
    DataAPI.SecIDGet(...)
    secID	ticker	secShortName	cnSpell	exchangeCD	assetClass	listStatusCD	listDate	transCurrCD	ISIN	partyID
    
    000001.ZICN	000001	上证综指	SZZZ	XSHG	IDX	L	1991-07-15	CNY	CNM000000019	17764
    399001.ZICN	399001	深证成指	SZCZ	XSHE	IDX	L	1995-01-23	CNY	CNM0000000C8	17765
    399006.ZICN	399006	创业板指	CYBZ	XSHE	IDX	L	2010-06-01	CNY	NaN	17765
    000300.ZICN	000300	沪深300	HS300	XSHG	IDX	L	2005-04-08	CNY	CNM0000001Y0	17768
    000016.ZICN	000016	上证50	SZ50	XSHG	IDX	L	2004-01-02	CNY	CNM0000001P8	17764
    000905.ZICN	000905	中证500	ZZ500	XSHG	IDX	L	2007-01-15	CNY	CNM0000007J8	17768
    
    """
    
    PB_PE.loc[date] = [sum(data['LFLO'] * data['PB']), sum(data['LFLO'] * data['PE']),MktIdxdGet_1.iloc[0,0],MktIdxdGet_1.iloc[1,0],MktIdxdGet_1.iloc[2,0],MktIdxdGet_1.iloc[3,0],MktIdxdGet_1.iloc[4,0],MktIdxdGet_1.iloc[5,0]]
    
# 增加2列
PB_PE['average PB(LFLO)'] = PB_PE['total PB(LFLO)'].mean()
PB_PE['average PE(LFLO)'] = PB_PE['total PE(LFLO)'].mean()
```

```python
PB_PE[['shangzheng','shenzheng','chuangye','hs300','shzh50','zhzh500']].plot(figsize=(25,7))

PB_PE_Relative = pd.DataFrame(index=Dates, data=0, columns=['shangzheng','shenzheng','chuangye','hs300','shzh50','zhzh500'])
PB_PE_Relative['shangzheng']=PB_PE['shangzheng']/sum(PB_PE['shangzheng'])
PB_PE_Relative['shenzheng']=PB_PE['shenzheng']/sum(PB_PE['shenzheng'])
PB_PE_Relative['chuangye']=PB_PE['chuangye']/sum(PB_PE['chuangye'])
PB_PE_Relative['hs300']=PB_PE['hs300']/sum(PB_PE['hs300'])
PB_PE_Relative['shzh50']=PB_PE['shzh50']/sum(PB_PE['shzh50'])
PB_PE_Relative['zhzh500']=PB_PE['zhzh500']/sum(PB_PE['zhzh500'])
PB_PE_Relative['average PE(LFLO)']=PB_PE['average PE(LFLO)']
PB_PE_Relative[['shangzheng','shenzheng','chuangye','hs300','shzh50','zhzh500','average PE(LFLO)']].plot(secondary_y=['average PE(LFLO)'],figsize=(25,7))

PB_PE[['total PB(LFLO)','average PB(LFLO)','zhzh500']].plot(secondary_y=['zhzh500'],figsize=(25,7))
PB_PE[['total PE(LFLO)','average PE(LFLO)','zhzh500']].plot(secondary_y=['zhzh500'],figsize=(25,7))

PB_PE[['total PB(LFLO)','average PB(LFLO)','shzh50']].plot(secondary_y=['shzh50'],figsize=(25,7))
PB_PE[['total PE(LFLO)','average PE(LFLO)','shzh50']].plot(secondary_y=['shzh50'],figsize=(25,7))
```
