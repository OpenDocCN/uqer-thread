# 分享一个 导出数据到本地的方法

### #好像有些人喜欢从优矿上拿数据，然后下载到自己电脑作分析。
### #我找到一个比较便捷的方法分享给大家 使用 pickle python自带的一个方法

```python
def SaveVarToFile(Var,filename):#保存变量Var到文件file 一个文件暂时只能保存一个Var 文件后缀为TXT
    import cPickle as pickle   
    write(filename,pickle.dumps(Var), encoding = 'utf-8', mode = 'w')
    return filename
def LoadVarFromFile(filename):#从文件file读取var  一个文件暂时只能保存一个Var
    import cPickle as pickle 
    p2=str(read(filename,encoding = 'utf-8'))
    Var = pickle.loads(p2)  
    return Var

#测试
#读取指数
data = DataAPI.MktIdxdGet(tradeDate=u"",indexID=u"",ticker=u"000001",beginDate='19910101',endDate='20160327',field=u"preCloseIndex,openIndex,tradeDate",pandas="1")
SaveVarToFile(data,'Indexlist.txt')#保存文件里，就可以下载到电脑使用了

#电脑端读取数据
LoadVarFromFile('Indexlist.txt')
```
