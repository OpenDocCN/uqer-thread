```python
raw_data = DataAPI.FundNavGet(secID=u"000001.OFCN",ticker=u"000001",dataDate=u"",beginDate=u"20171201",endDate=u"20191201",partyID="",partyShortName="",field=u"endDate,NAV",pandas="1")
# print(raw_data)
data_list = [enddate for enddate in raw_data.endDate]
# print(data_list)
nav_list = [nav for nav in raw_data.NAV]
data_dict = {}
#近两年的
tdata_list = ['2017-12','2018-01','2018-02','2018-03','2018-04','2018-05','2018-06','2018-07','2018-08','2018-09','2018-10','2018-11','2018-12','2019-01','2019-02','2019-03','2019-04','2019-05','2019-06','2019-07','2019-02','2019-09','2019-10','2019-11']
a = []
b = []

#计算一个周期内的的最回撤率
def max_count(min,max):
    return 1-min/max

for i in range(490):
    data_dict[data_list[i]] = nav_list[i] 
# print(data_dict)
for x in tdata_list:
    for i,j in data_dict.items():
        if x in i:
            a.append(j)
    a.sort()
    # print(a)
    min = a[0]
    max = a.pop()
    result = max_count(min,max)
    b.append(result)
    a = []
b.sort()
print('2年最大回撤率：%s'%b.pop())
        


    
```
