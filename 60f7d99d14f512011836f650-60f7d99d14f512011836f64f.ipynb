```python
#2.5numpy的索引
import numpy as np
A=np.arange(3,15)
print(A)
print(A[3])#输出一维数组A中第3个位置的值，注意位置是从0开始的

A=np.arange(3,15).reshape((3,4))
print(A)
print(A[2])#输出二维数组A中第2行的值
print(A[1][1])#输出A中第一行第一列的值
print(A[1,1])#和上一种形式结果一致
print(A[2,:])#输出A中第二行所有的值
print(A[:,1])#输出A中第一列所有的值
print(A[1,1:2])#输出A中第一列，第一行到第二行的值，注意含头不含尾
print(A[1,1:3])

A=np.arange(3,15).reshape((3,4))
print"A："
print(A)
print"迭代A中的行："
for row in A:
    print(row)
print"迭代A中的列："
for column in A.T:
    print(column)
    
A=np.arange(3,15).reshape((3,4))
print"A："
print(A)
print"A.flat："#返回的是一个迭代器？？？
print(A.flat)
print"A.flatten："#输出的是一个被改变的array，一维数组
print(A.flatten())
for item in A.flat:#输出被改变后的A的每一项
    print(item)
    
#2.6numpy的array合并
import numpy as np
A=np.array([1,1,1])
B=np.array([2,2,2])
C=np.vstack((A,B))#上下合并vertical stack
print C
print(A.shape,C.shape)#A只是一个序列，C是2行3列的矩阵
D=np.hstack((A,B))#左右合并horizontal stack
print(A.shape,D.shape)

import numpy as np
A=np.array([1,1,1])[:,np.newaxis]#从一行array转为一列之后horizontal stack合并
B=np.array([2,2,2])[:,np.newaxis]
print A
print B
C=np.hstack((A,B))
print C
D=np.concatenate((A,B,B,A),axis=0)#concatenate可以选择在哪一个维度合并，axis=0则是纵向合并，而stack只能单个维度合并
print(D)
E=np.concatenate((A,B,B,A),axis=1)
print(E)

#2.7numpy的array分割
print"#numpy的array分割"
import numpy as np
A=np.arange(12).reshape((3,4))
print"A:"
print(A)
print"将A等项分割"
print(np.split(A,2,axis=1))#将A按列切割成2块
print(np.split(A,3,axis=0))#将A按行切割成3块
print"将A不等项分割"
print(np.array_split(A,3,axis=1))#将A按列不等项切割成3块
print(np.vsplit(A,3))#vertical split可以完成不等项的垂直分割
print(np.hsplit(A,2))#horizontal split可以完成不等项的水平分割

#2.8numpy的 copy & deep copy
print"#numpy的 copy & deep copy"
import numpy as np
a=np.arange(4)
print a
b=a
c=a
d=b
a[0]=11
print a
print"b:"#改变a的第0项，b也跟着改变
print b
b is a#返回true，说明b跟着a改变
c is a#改变a，
d is a

d[1:3]=[22,33]#反之，修改了d，abc都会改变，这四个数组通过等号关联在了一起，改变其中一个就等于改变所有
print"d:"
print d
print a
print b
print c

a=np.arange(4)
b=a.copy()#deep copy也就是把a的值赋给了b，但是没有把b关联给a，之后改变a，不改变b
print a
print b
print"深度赋值deep copy"
a[3]=44
print a
print b

```
