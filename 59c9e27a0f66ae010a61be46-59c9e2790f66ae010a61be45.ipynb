# 事件研究——财报对分析师评级上调事件的影响

分析师评级上调往往被认为是非常正面的事件，在事件发生后长期来看有不错的超额收益。而实际上评级上调事件带来超额收益的原因至少有两个层面。第一个层面自然是分析师的能力以及影响力，这也是大家普遍认同的。而第二个层面是其他事件的影响，当公司发生某些重大利好事件时，比如发布了超预期的财报，那么分析师就更容易对此股票做出评级上调。在这些重大利好事件中，财报是频繁发生且代表性较强的事件，本篇文章以评级报告发布前一个月是否有财报发布为标准，将评级报告分为两组，以比较财报对评级上调事件的影响。

```python
import pandas as pd
import numpy as np
from quartz_extensions.EventAnalysis import data_infer,event_service,normalize_event
import matplotlib.pyplot as plt
from CAL.PyCAL import *    # CAL.PyCAL中包含font
font.set_size(15)
```

```python
从优矿中取出2011年以来所有A股分析师个股报告的数据，并清洗数据。
```

```python
def clean_data(report):
    report['event_date'] = pd.to_datetime(pd.to_datetime(report['intoTime']).dt.date)
    report = report.drop('intoTime', axis=1)
    report = report.sort_values(['event_date', 'ticker'])
    report.index = range(len(report))
    report = report.loc[report['event_date'] >= pd.to_datetime('20110101')]
    return report

def get_report():
    report_list = []
    fi_statement_list = []
    tickera = DataAPI.EquGet(equTypeCD=u"A",pandas="1")
    for ticker in np.sort(tickera['ticker']):
        stock_report=DataAPI.GG.derReportResGGGet (secID=u"",ticker=ticker,beginDate=u"20110101",\
                        endDate=u"20170920",field=u"ticker,intoTime,scoreFlag,priceCurrent",pandas="1")
        stock_fi_statement = DataAPI.FdmtISGet(ticker=ticker,publishDateEnd=u"20170920",publishDateBegin=u"20100101",\
                                               field=u"ticker,publishDate",pandas="1")
        report_list.append(stock_report)
        fi_statement_list.append(stock_fi_statement)
    report = pd.concat(report_list, axis=0)
    report = clean_data(report)
    report.to_excel('analyst_report_clean.xlsx')
    fi_statement = pd.concat(fi_statement_list, axis=0)
    fi_statement.to_excel('financial_statement.xlsx')
    return report

report = get_report()
print(report.head())
```

由于股价低于1元的股票受到滑点影响较大，最小的交易单位1分钱就会对他产生 1% 收益率的影响，所以我们剔除这些股价低于1元的股票。其次，剔除重复的事件，即对于同一天同一只股票最多只算一个事件。最后剔除事件后一天即买入日停牌、涨停、跌停的股票。

```python
report = pd.read_excel('analyst_report_clean.xlsx')
report = report.loc[report['priceCurrent'] >= 1]
report['ticker'] = report['ticker'].astype(str).str.zfill(6)
report = report.sort_values(['event_date', 'ticker'])
#取出评级上调事件
event = report.loc[report['scoreFlag'] > 0, ['ticker', 'event_date']]
print(event.head())
```

用事件推断器包装事件， BASE_DATE统一定为事件日下一个交易日

```python
event['event_date'] = pd.to_datetime(event['event_date']).dt.strftime('%Y-%m-%d')
event = normalize_event(event, publish_date_field='event_date', security_field='ticker')
data_infer.build(event)
event = event.dropna(subset=['_PUBLISH_DATE'])
event = data_infer.build(event).append('BASE_DATE',adjust='last', step=1).get()
buy_price = 'closePrice'
event = data_infer.build(event).append(buy_price).append('limitUpPrice').append('limitDownPrice').append('isHalt').get()
event = event.loc[(event[buy_price] < event['limitUpPrice']) & (event[buy_price] > event['limitDownPrice'])]
event = event.loc[event['isHalt'] == u'未停牌']
event = event.drop_duplicates(['ticker', 'BASE_DATE'])
print(event.head())
print(len(event))
```

统计每个月份评级上调事件发生的频率，发现3、4、8月事件个数明显多于其他月份，而这几个月份正是年报和半年报发布频繁的日期。

```python
event['month'] = pd.to_datetime(event['event_date']).dt.month
monthly_num = event.groupby('month')['ticker'].count()
monthly_num = monthly_num.rename(columns={'ticker':'monthly_num'})
fig, ax = plt.subplots(1, 1)
monthly_num.plot(ax=ax, kind='bar')
ax.set_title(u'评级上调事件数量的月份分布', fontproperties=font)
ax.set_xlabel(u'月份', fontproperties=font)
ax.set_ylabel(u'事件数量', fontproperties=font)
```

进一步，我们按照评级上调前一个月是否有财报发布区分评级上调事件，并统计两组事件的个数， 可以看到有财报次数明显大于无财报次数。而实际上财报后一个月这个定义的时间段在一年中只占了1/3。所以我们认为分析师更倾向于在财报发布后一段时间对公司进行评级上调。

```python
def merge_asof(left_df, right_df, groupby_col, left_on, right_on, allow_exact_matches, tolerance, right_date_col):
    left_df[left_on] = pd.to_datetime(left_df[left_on])
    right_df[right_on] = pd.to_datetime(right_df[right_on])
    left_df_list = []
    for group_name in left_df[groupby_col].unique():
        left_group = left_df.loc[left_df[groupby_col] == group_name].copy()
        left_group[right_date_col] = np.nan 
        right_group = right_df[right_df[groupby_col] == group_name].copy()
        for idx in left_group.index:
            left_date = left_group.loc[idx, left_on]
            if allow_exact_matches:
                right_dates = right_group.loc[right_group[right_on] <= left_date, right_on]
            else:
                right_dates = right_group.loc[right_group[right_on] < left_date, right_on]
            if len(right_dates) > 0:
                latest_date = right_dates.iloc[-1]
                if (left_date - latest_date).days <= tolerance:
                    left_group.loc[idx, right_date_col] = latest_date
        left_df_list.append(left_group)         
    left_df = pd.concat(left_df_list, axis=0)
    left_df[left_on] = left_df[left_on].dt.strftime('%Y-%m-%d') 
    return left_df

#获取财报发布日数据
fi_statement = pd.read_excel('financial_statement.xlsx')
fi_statement['ticker'] = fi_statement['ticker'].astype(str).str.zfill(6)
fi_statement = fi_statement.sort_values(['publishDate', 'ticker'])
#匹配事件与财报
event = event.sort_values(['BASE_DATE', 'ticker'])
event = merge_asof(event, fi_statement, groupby_col='ticker', left_on='BASE_DATE', right_on='publishDate', allow_exact_matches=False,                               tolerance=30, right_date_col='has_statement')
event.loc[~event['has_statement'].isnull(), 'has_statement'] = 1
event['has_statement'] = event['has_statement'].fillna(0)
event = event.sort_values(['BASE_DATE', 'ticker'])

print('评级上调事件发布前一个月有财报的次数： %s' %(len(event[event['has_statement'] == 1])))
print('评级上调事件发布前一个月无财报的次数： %s' %(len(event[event['has_statement'] == 0])))
```

我们再来比较前一个月有无财报发布对事件发生后累积超额收益的影响

```python
def get_transfer_window(event):
    transfer_window_gains = event_service.summary_window_gains(event, range(-20,60), '000300', start_point=-20,                                                                              ret_type = 'c2c')
    transfer_window_gains['ACTIVE_CUMSUM_GAINS'] = transfer_window_gains['ABS_CUMSUM_GAINS']-transfer_window_gains['BASE_CUMSUM_GAINS']+1
    active_gains = transfer_window_gains['ACTIVE_CUMSUM_GAINS'].iloc[20:]
    active_gains = active_gains / active_gains.iloc[0] - 1
    return active_gains
```

```python
event = event.loc[pd.to_datetime(event['BASE_DATE']) <= pd.to_datetime('20170601')]
event_has_statement = event.loc[event['has_statement'] == 1] 
event_no_statement = event.loc[event['has_statement'] != 1]   
```

```python
active_gains_post_event_statement = get_transfer_window(event_has_statement)
active_gains_post_event_no_statement = get_transfer_window(event_no_statement)
active_gains_post_event = pd.concat([active_gains_post_event_statement, active_gains_post_event_no_statement], axis=1)
active_gains_post_event.columns = ['has_statement', 'no_statement']
```

通过画图可以发现，前一个月无财报发布的分析师评级上调事件, 在事件发生后的超额收益，明显高于前一个月有财报发布的分析师评级上调事件，并且两者差距随持有时间增加不断变大。

```python
#事件发生后累积超额收益（相对沪深300）对比
fig, ax = plt.subplots(1, 1)
active_gains_post_event.plot(ax=ax)
ax.set_title(u'事件发生后累积超额收益率（相对沪深300）对比', fontproperties=font)
ax.set_xlabel(u'持有交易日数', fontproperties=font)
ax.set_ylabel(u'累积超额收益率', fontproperties=font)
```

这里我们只展示了超额收益的情况，实际上经过回测，剔除了行业、市值等因素影响后的特异性收益也出现了同样的现象。发生这类现象可能是因为良好的财报是一种广为人知且被大家普遍认同的利好，价格的反应也就更加迅速和充分，所以前一个月有财报发布的评级上调事件后期涨幅相对较小。

分析师评级上调事件只是一个例子，类似的，我们可以发现许多事件之间有相互的影响。理清这些相互影响，挖掘事件本身带来的收益，方能带来更好的回报。
