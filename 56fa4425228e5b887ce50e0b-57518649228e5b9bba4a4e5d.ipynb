```python
# built-in
import time
import operator
import collections
import datetime as dt
from functools import wraps

# 3-party
from CAL.PyCAL import *
import pandas as pd
pd.options.display.max_columns = 100
pd.options.display.max_rows = 300
import numpy as np
import seaborn
from matplotlib import pylab
import matplotlib.pyplot as plt
```

```python
### helpers and wrappers

def wrapper_func_cache(func):
    @wraps(func)
    def wrapped(*args, **kwargs):
        if not hasattr(func, '__cache_result'):
            func.__cache_result = {}
        
        hash_key = [str(i) for i in args] + [str(j) for j in kwargs.iteritems()]
        cache_key = hash(tuple(hash_key))
        
        if cache_key not in func.__cache_result:
            func.__cache_result[cache_key] = func(*args, **kwargs)
            
        return func.__cache_result[cache_key]
    return wrapped

@wrapper_func_cache
def const_trade_calendar():
    data = DataAPI.TradeCalGet(exchangeCD=u"XSHG,XSHE", beginDate=u"20050101", 
                               endDate=u"", field=['calendarDate', 'isOpen'])
    data = data[data.isOpen == 1]

    return list(data.calendarDate)

@wrapper_func_cache
def const_pre_trade_day(day):
    trade_days = const_trade_calendar() 
    index = trade_days.index(day)
    
    return trade_days[index - 1]
```

### step 3: 编写自己的归因函数

```python
def extract_order(record):
    str_order = str(record)
    list_order = str_order[str_order.index('(') + 1 : str_order.index(')')].split(",")
    list_order = [i.strip() for i in list_order]
    
    def str2record(_str):
        """
        'order_time: 2013-01-08 09:30'  -> ['order_time', '2013-01-08 09:30']
        """
        if _str.startswith('order_time') or _str.startswith('filled_time'):
            return [_str[0 : _str.index(":")], _str[_str.index(":") + 1 : ]]
        else:
            return [i.strip() for i in _str.split(':')]
    
    list_order = [str2record(i) for i in list_order]
    
    return list_order

def extract_blotter(blotter):
    all_orders = []
    for i in blotter:
        all_orders.extend(i)
    
    all_records = {'order_time': [], 'symbol': [], 'direction': [], 'order_amount': [], 'state': [],
                   'filled_time': [], 'filled_amount': [], 'transact_price': []}
    for i in all_orders:
        tmp = extract_order(i)
        for term, value in tmp:
            if term in all_records:
                all_records[term].append(value)
    
    return all_records

def attribution(bt, perf):
    """回测归因报告
    
    return: blotter, report，rebalance   
    """
    report = pd.DataFrame(columns=['项目', '数据'])
    detail_report = pd.DataFrame(columns=['项目', '数据'])
    term, value = [], []
    
    blotter = extract_blotter(bt.blotter)
    blotter = pd.DataFrame.from_records(blotter)
    blotter.direction = blotter.direction.astype(float)
    blotter.filled_amount = blotter.filled_amount.astype(float)
    blotter.transact_price = blotter.transact_price.astype(float)
    blotter.order_amount = blotter.order_amount.astype(float)
    
    # 交易频率
    trade_days = sorted(list(set(list(blotter.order_time))))
    tstart, tend = trade_days[0][:11].strip(), trade_days[-1][:11].strip()
    full_trade_days = lib.trade_range(tstart, tend)
    term.extend(['回测时间', '回测期间总交易日', '策略交易日', '交易频率'])
    value.extend(['{} -> {}'.format(tstart, tend), len(full_trade_days), len(trade_days), 
                                    round(len(trade_days)*1./len(full_trade_days), 2)])
    
    # 订单状态
    state_counter = collections.Counter(list(blotter.state))
    total = sum(state_counter.values())
    for i in state_counter:
        term.append('订单状态: ' + i)
        value.append('{:<4} | {}'.format(state_counter[i], round(state_counter[i]*1./total, 3)))
    
    # 证券分布
    tickers = collections.Counter(list(blotter.symbol))
    total = sum(tickers.values())
    term.extend(['交易证券支数', '交易次数'])
    value.extend([len(tickers), total])
    
    # 持仓周期四分位统计，可以画图
    
    # 日收益，周收益分布，个股资金流策略，在分钟线和日线回测时，2014年performance差异有点大
    
    # 持仓股票分类：市值，pe，pb
    
    # 回测报表
    report['项目'], report['数据'] = term, value
    print report.to_html()
    
    return blotter, report, rebalance
```

### step 3: 每一个策略回测结束后都会有两个回测变量，bt，perf，我们就用这两个回测变量来归因，详细文档看这里：[bt,perf](https://uqer.datayes.com/help/faqApi/#策略结果)
