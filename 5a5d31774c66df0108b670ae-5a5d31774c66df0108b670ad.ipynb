```python
#获取目前上市国债ID
ID=DataAPI.SecIDGet(assetClass=u"B",field=['secID','secShortName','listStatusCD'],pandas='1')
ID=ID[ID['listStatusCD']=='L']
ID=ID['secID'][:13]
#获取债券基本信息
df=DataAPI.BondGet(secID=ID,field=['secID','couponTypeCD','cpnFreqCD','par','coupon','maturityDate'],pandas="1").set_index('secID')
df=df[df['couponTypeCD']=='FIXED']
#获取债券收盘价
datetoday='2018-01-11'
import pandas as pd
price=DataAPI.MktBonddGet(tradeDate=datetoday,secID=ID,field=['secID','closePrice'],pandas="1").set_index('secID')
pricedf=pd.concat([price,df],axis='1')
pricedf.dropna(axis=0,how='any',inplace=True)
```

```python
# 计算债券到期收益率
import numpy as np
import pandas as pd
from datetime import datetime
from matplotlib import pyplot as plt
global YTMdf
def calc_YTM(data):
    YTM=[]
    period=[]
    int_ac=[]
    per=[]
    for bd in data.index:
        freq=int(data['cpnFreqCD'][bd][:1])
        pmt=data['coupon'][bd]/freq
        pv=-data['closePrice'][bd]
        fv=data['par'][bd]
        date1=datetime.strptime(data['maturityDate'][bd],'%Y-%m-%d')
        date2=datetime.today()
        nper=(date1-date2).days/182.5
        a=np.rate(nper, pmt, pv, fv,when='end', guess=0.03, tol=1e-06, maxiter=100)
        a=a*freq
        YTM.append(a)
        period.append(nper)
        days=(int(nper)+1-nper)
        b=data['coupon'][bd]/freq*days
        int_ac.append(b)
        per.append(freq)
    return YTM,period,int_ac,per
YTM,period,int_ac,per=calc_YTM(pricedf)
YTMdf=pd.DataFrame(YTM,index=pricedf.index,columns=['YTM'])
YTMdf['period']=period
YTMdf['int_ac']=int_ac
YTMdf['coupon']=pricedf['coupon']
YTMdf['par']=pricedf['par']
YTMdf['closePrice']=pricedf['closePrice']
YTMdf['per']=per
plt.scatter(YTMdf['period'],YTMdf['YTM'])
YTMdf.to_csv('YTMdf.csv')
```

```python
from scipy import optimize as op
from numpy import exp

# nss-model 计算每期的即期利率
def ns_model(r, x):
    if r==0:
        return 0.0283
    a = r / x[3]
    b = 1.0 - exp(-a)
    spot_rate = x[0] + x[1] * (b / a) + x[2] * (b / a - exp(-a))
    return spot_rate


#根据即期利率计算债券理论价格
def bond_price(per, int_ac, r, coupon, par, YTM, closePrice, x):
    p = 0
    duration = 0
    for i in range(int(r)):
        t = (i + 1.0) / 2.0
        spot_rate = ns_model(t, x)
        dis_rate = exp(-t * spot_rate)
        cf = coupon / per
        p = p + cf * dis_rate
        duration = duration + cf*t / ((1 + YTM / per) ** t)
    p = p + par * dis_rate - int_ac
    duration = (duration + par*t / ((1 + YTM / per) ** t) / closePrice * (1.0 / (1 + YTM / per)))
    return p, duration


# 设置目标函数
def objective(x):
    bprice = []
    dur = []
    for i in range(len(YTMdf.index)):
        p, duration = bond_price(YTMdf['per'][i], YTMdf['int_ac'][i], YTMdf['period'][i], YTMdf['coupon'][i],
                                 YTMdf['par'][i], YTMdf['YTM'][i], YTMdf['closePrice'][i], x)
        bprice.append(p)
        dur.append(duration)
    df = pd.DataFrame(YTMdf['closePrice'], index=YTMdf.index, columns=['closePrice'])
    df['p'] = bprice
    df['dur'] = dur
    df['w'] = 1.0 / df['dur'] / sum(1.0 / df['dur'])
    #根据久期赋权
#    df['dif'] = df['w'] * (df['p'] - df['closePrice']) * (df['p'] - df['closePrice'])
    #等权重
    df['dif'] = 0.125* (df['p'] - df['closePrice']) * (df['p'] - df['closePrice'])
    return sum(df['dif'])

#限制条件
def cons1(x):
    return x[0]


def cons2(x):
    return x[0] + x[1]

```

```python
import numpy as np
import pandas as pd
#利用op.fmin_cobyla进行优化求解，这里本人对各种优化方法不是很了解，希望对此有了解的人能够指出更合适的优化方法。
out = op.fmin_cobyla(objective, [0.05,-0.02,-0.01,1], [cons1, cons2])
print(out)
#画图
timespace=[0,0.08333,0.1666,0.25,0.5,0.75,1,3,5,7,10,15,20,30,40,50]
curvedf=pd.DataFrame(0,index=timespace,columns=['rate'])
rate=[]
for i in timespace:
    s_rate=ns_model(i, out)
    rate.append(s_rate)
curvedf['rate']=rate
#实际收益率曲线r_rate
r_rate=[2.5869,3.0995,3.2709,3.3386,3.5514,3.5744,3.5827,3.6619,3.8598,3.8856,3.9337,4.1933,4.232,4.3526,4.3926,4.403]    
curvedf['r_rate']=r_rate
curvedf['r_rate']=curvedf['r_rate']/100.0
print(curvedf)
curvedf.plot(figsize=(16,10))
```
