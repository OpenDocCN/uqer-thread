```python
import pandas as pd
import numpy as np
import math
from datetime import datetime, timedelta
from dateutil.parser import parse
import matplotlib.pyplot as plt
import talib as ta 
import time
from CAL.PyCAL import * 
import matplotlib.colors as colors
import matplotlib.cm as cmx
import random as rd
import requests
import json
import re
from dateutil import rrule
import seaborn as sns
import quartz_extensions.FundAnalysis.analysis as fa

font.set_size(22)
small_font = font.copy()
small_font.set_size(14)

from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# -*- coding:utf-8 -*- 

def get_recent_date(i):
    #获取最近开市日期
    start_date='20200901'
    # starttime = datetime.now()
    mytoday=datetime.today().strftime('%Y%m%d')
    mydate=DataAPI.TradeCalGet(exchangeCD=u"XSHG",beginDate=start_date,endDate=mytoday,isOpen=u"1",field=u",isOpen,calendarDate,prevTradeDate",pandas="1")
    # n=-3 
    n=-1 
    if mytoday==mydate.iloc[-1]["calendarDate"] and int(datetime.now().strftime("%H"))<22:
    #如果开始当天时间超过晚上十点，取当天数据
        n=n-1  
    # recent_date=mydate.iloc[-4-i]['calendarDate']
    recent_date=mydate.iloc[n-i]['calendarDate']
    return recent_date


#报告期
before_16days=get_recent_date(16)
index_month=before_16days.find('-')+1
year_before_16days=before_16days[:index_month-1]

if parse(get_recent_date(16))>=parse(year_before_16days+"-"+"01-01") and parse(get_recent_date(16))<parse(year_before_16days+"-"+"03-31"):
    myreportDate=str(int(year_before_16days)-1)+"-"+"12-31"
if parse(get_recent_date(16))>=parse(year_before_16days+"-"+"03-31") and parse(get_recent_date(16))<parse(year_before_16days+"-"+"06-30"):
    myreportDate=year_before_16days+"-"+"03-31"
if parse(get_recent_date(16))>=parse(year_before_16days+"-"+"06-30") and parse(get_recent_date(16))<parse(year_before_16days+"-"+"09-30"):
    myreportDate=year_before_16days+"-"+"06-30"
if parse(get_recent_date(16))>=parse(year_before_16days+"-"+"09-30") and parse(get_recent_date(16))<parse(year_before_16days+"-"+"12-31"):
    myreportDate=year_before_16days+"-"+"09-30"    

print("最近开市日期："),    
print(get_recent_date(0))
print("最近基金报告日期："),      
print(myreportDate)
# myreportDate="2021-12-31"

#myn=10,获取前十子基金，否则全部全部子基金
myn=11
#智能分类
n_cluster = 2
#显示数量
number_show=30
#投资本基金公司基金
data_inner = pd.DataFrame()

#------------------ 0级：核心工具函数 --------------------------------------------      
def RD(N,D=3):   return np.round(N,D)        #四舍五入取3位小数 
def RET(S,N=1):  return np.array(S)[-N]      #返回序列倒数第N个值,默认返回最后一个
def ABS(S):      return np.abs(S)            #返回N的绝对值
def MAX(S1,S2):  return np.maximum(S1,S2)    #序列max
def MIN(S1,S2):  return np.minimum(S1,S2)    #序列min
         
def MA(S,N):           #求序列的N日平均值，返回序列                    
    return pd.Series(S).rolling(N).mean().values     # pd.rolling_mean(S,N)  (Python2)

def REF(S, N=1):       #对序列整体下移动N,返回序列(shift后会产生NAN)    
    return pd.Series(S).shift(N).values  

def DIFF(S, N=1):      #前一个值减后一个值,前面会产生nan 
    return pd.Series(S).diff(N)  #np.diff(S)直接删除nan，会少一行

def STD(S,N):           #求序列的N日标准差，返回序列    
    return  pd.Series(S).rolling(N).std(ddof=0).values     

def IF(S_BOOL,S_TRUE,S_FALSE):          #序列布尔判断 res=S_TRUE if S_BOOL==True  else  S_FALSE
    return np.where(S_BOOL, S_TRUE, S_FALSE)

def SUM(S, N):            #对序列求N天累计和，返回序列    N=0对序列所有依次求和         
    return pd.Series(S).rolling(N).sum().values if N>0 else pd.Series(S).cumsum()    #pd.rolling_sum(S,N)  (Python2)

def HHV(S,N):             # HHV(C, 5)  # 最近5天收盘最高价        
    return pd.Series(S).rolling(N).max().values      # pd.rolling_max(S,N)  (Python2)

def LLV(S,N):             # LLV(C, 5)  # 最近5天收盘最低价     
    return pd.Series(S).rolling(N).min().values      # pd.rolling_min(S,N)  (Python2)

def EMA(S,N):             #指数移动平均,为了精度 S>4*N  EMA至少需要120周期       
    return pd.Series(S).ewm(span=N, adjust=False).mean().values    

def SMA(S, N, M=1):        #中国式的SMA,至少需要120周期才精确 (雪球180周期)    alpha=1/(1+com)
    return pd.Series(S).ewm(com=N-M, adjust=True).mean().values     

def AVEDEV(S,N):           #平均绝对偏差  (序列与其平均值的绝对差的平均值)   
    avedev=pd.Series(S).rolling(N).apply(lambda x: (np.abs(x - x.mean())).mean())    
    return avedev.values

def SLOPE(S,N,RS=False):    #返S序列N周期回线性回归斜率 (默认只返回斜率,不返回整个直线序列)
    M=pd.Series(S[-N:]);   poly = np.polyfit(M.index, M.values,deg=1);    Y=np.polyval(poly, M.index); 
    if RS: return Y[1]-Y[0],Y
    return Y[1]-Y[0]

#------------------   2级：技术指标函数(全部通过0级，1级函数实现） ------------------------------
def MACD(CLOSE,SHORT=12,LONG=26,M=9):            # EMA的关系，S取120日，和雪球小数点2位相同
    DIF = EMA(CLOSE,SHORT)-EMA(CLOSE,LONG);  
    DEA = EMA(DIF,M);      MACD=(DIF-DEA)*2
    return RD(DIF),RD(DEA),RD(MACD)

def KDJ(CLOSE,HIGH,LOW, N=9,M1=3,M2=3):         # KDJ指标
    RSV = (CLOSE - LLV(LOW, N)) / (HHV(HIGH, N) - LLV(LOW, N)) * 100
    K = EMA(RSV, (M1*2-1));    D = EMA(K,(M2*2-1));        J=K*3-D*2
    return K, D, J

def RSI(CLOSE, N=24):      
    DIF = CLOSE-REF(CLOSE,1) 
    return RD(SMA(MAX(DIF,0), N) / SMA(ABS(DIF), N) * 100)  

def WR(CLOSE, HIGH, LOW, N=10, N1=6):            #W&R 威廉指标
    WR = (HHV(HIGH, N) - CLOSE) / (HHV(HIGH, N) - LLV(LOW, N)) * 100
    WR1 = (HHV(HIGH, N1) - CLOSE) / (HHV(HIGH, N1) - LLV(LOW, N1)) * 100
    return RD(WR), RD(WR1)

def BIAS(CLOSE,L1=6, L2=12, L3=24):              # BIAS乖离率
    BIAS1 = (CLOSE - MA(CLOSE, L1)) / MA(CLOSE, L1) * 100
    BIAS2 = (CLOSE - MA(CLOSE, L2)) / MA(CLOSE, L2) * 100
    BIAS3 = (CLOSE - MA(CLOSE, L3)) / MA(CLOSE, L3) * 100
    return RD(BIAS1), RD(BIAS2), RD(BIAS3)

def BOLL(CLOSE,N=20, P=2):                       #BOLL指标，布林带    
    MID = MA(CLOSE, N); 
    UPPER = MID + STD(CLOSE, N) * P
    LOWER = MID - STD(CLOSE, N) * P
    return RD(UPPER), RD(MID), RD(LOWER)    

def PSY(CLOSE,N=12, M=6):  
    PSY=COUNT(CLOSE>REF(CLOSE,1),N)/N*100
    PSYMA=MA(PSY,M)
    return RD(PSY),RD(PSYMA)

def CCI(CLOSE,HIGH,LOW,N=14):  
    TP=(HIGH+LOW+CLOSE)/3
    return (TP-MA(TP,N))/(0.015*AVEDEV(TP,N))
        
def ATR(CLOSE,HIGH,LOW, N=20):                    #真实波动N日平均值
    TR = MAX(MAX((HIGH - LOW), ABS(REF(CLOSE, 1) - HIGH)), ABS(REF(CLOSE, 1) - LOW))
    return MA(TR, N)

def BBI(CLOSE,M1=3,M2=6,M3=12,M4=20):             #BBI多空指标   
    return (MA(CLOSE,M1)+MA(CLOSE,M2)+MA(CLOSE,M3)+MA(CLOSE,M4))/4    

def DMI(CLOSE,HIGH,LOW,M1=14,M2=6):               #动向指标：结果和同花顺，通达信完全一致
    TR = SUM(MAX(MAX(HIGH - LOW, ABS(HIGH - REF(CLOSE, 1))), ABS(LOW - REF(CLOSE, 1))), M1)
    HD = HIGH - REF(HIGH, 1);     LD = REF(LOW, 1) - LOW
    DMP = SUM(IF((HD > 0) & (HD > LD), HD, 0), M1)
    DMM = SUM(IF((LD > 0) & (LD > HD), LD, 0), M1)
    PDI = DMP * 100 / TR;         MDI = DMM * 100 / TR
    ADX = MA(ABS(MDI - PDI) / (PDI + MDI) * 100, M2)
    ADXR = (ADX + REF(ADX, M2)) / 2
    return PDI, MDI, ADX, ADXR  

def TAQ(HIGH,LOW,N):                               #唐安奇通道(海龟)交易指标，大道至简，能穿越牛熊
    UP=HHV(HIGH,N);    DOWN=LLV(LOW,N);    MID=(UP+DOWN)/2
    return UP,MID,DOWN

def KTN(CLOSE,HIGH,LOW,N=20,M=10):                 #肯特纳交易通道, N选20日，ATR选10日
    MID=EMA((HIGH+LOW+CLOSE)/3,N)
    ATRN=ATR(CLOSE,HIGH,LOW,M)
    UPPER=MID+2*ATRN;   LOWER=MID-2*ATRN
    return UPPER,MID,LOWER       
  
def TRIX(CLOSE,M1=12, M2=20):                      #三重指数平滑平均线
    TR = EMA(EMA(EMA(CLOSE, M1), M1), M1)
    TRIX = (TR - REF(TR, 1)) / REF(TR, 1) * 100
    TRMA = MA(TRIX, M2)
    return TRIX, TRMA

def VR(CLOSE,VOL,M1=26):                            #VR容量比率
    LC = REF(CLOSE, 1)
    return SUM(IF(CLOSE > LC, VOL, 0), M1) / SUM(IF(CLOSE <= LC, VOL, 0), M1) * 100

def EMV(HIGH,LOW,VOL,N=14,M=9):                     #简易波动指标 
    VOLUME=MA(VOL,N)/VOL;       MID=100*(HIGH+LOW-REF(HIGH+LOW,1))/(HIGH+LOW)
    EMV=MA(MID*VOLUME*(HIGH-LOW)/MA(HIGH-LOW,N),N);    MAEMV=MA(EMV,M)
    return EMV,MAEMV


def DPO(CLOSE,M1=20, M2=10, M3=6):                  #区间震荡线
    DPO = CLOSE - REF(MA(CLOSE, M1), M2);    MADPO = MA(DPO, M3)
    return DPO, MADPO

def BRAR(OPEN,CLOSE,HIGH,LOW,M1=26):                 #BRAR-ARBR 情绪指标  
    AR = SUM(HIGH - OPEN, M1) / SUM(OPEN - LOW, M1) * 100
    BR = SUM(MAX(0, HIGH - REF(CLOSE, 1)), M1) / SUM(MAX(0, REF(CLOSE, 1) - LOW), M1) * 100
    return AR, BR

def DMA(CLOSE,N1=10,N2=50,M=10):                     #平行线差指标  
    DIF=MA(CLOSE,N1)-MA(CLOSE,N2);    DIFMA=MA(DIF,M)
    return DIF,DIFMA

def MTM(CLOSE,N=12,M=6):                             #动量指标
    MTM=CLOSE-REF(CLOSE,N);         MTMMA=MA(MTM,M)
    return MTM,MTMMA

def MASS(HIGH,LOW,N1=9,N2=25,M=6):                   # 梅斯线
    MASS=SUM(MA(HIGH-LOW,N1)/MA(MA(HIGH-LOW,N1),N1),N2)
    MA_MASS=MA(MASS,M)
    return MASS,MA_MASS
  
def ROC(CLOSE,N=12,M=6):                             #变动率指标
    ROC=100*(CLOSE-REF(CLOSE,N))/REF(CLOSE,N);    MAROC=MA(ROC,M)
    return ROC,MAROC  

def EXPMA(CLOSE,N1=12,N2=50):                        #EMA指数平均数指标
    return EMA(CLOSE,N1),EMA(CLOSE,N2);

def OBV(CLOSE,VOL):                                  #能量潮指标
    return SUM(IF(CLOSE>REF(CLOSE,1),VOL,IF(CLOSE<REF(CLOSE,1),-VOL,0)),0)/10000

def ASI(OPEN,CLOSE,HIGH,LOW,M1=26,M2=10):            #振动升降指标
    LC=REF(CLOSE,1);      AA=ABS(HIGH-LC);     BB=ABS(LOW-LC);
    CC=ABS(HIGH-REF(LOW,1));   DD=ABS(LC-REF(OPEN,1));
    R=IF( (AA>BB) & (AA>CC),AA+BB/2+DD/4,IF( (BB>CC) & (BB>AA),BB+AA/2+DD/4,CC+DD/4));
    X=(CLOSE-LC+(CLOSE-OPEN)/2+LC-REF(OPEN,1));
    SI=16*X/R*MAX(AA,BB);   ASI=SUM(SI,M1);   ASIT=MA(ASI,M2);
    return ASI,ASIT   
#------------------ 0级：核心工具函数 --------------------------------------------  

def analysis_stock(ticker):
    myticker=ticker
    spdb=DataAPI.FundNavGet(secID=u"",ticker=myticker,dataDate=u"",beginDate=u"20101001",endDate=u"",partyID="",partyShortName="",field=u"",pandas="1")
    spdb['tradeDate']=spdb['endDate']
    spdb['closeIndex']=spdb['ADJUST_NAV']
    spdb['highestIndex']=spdb['closeIndex']
    spdb['lowestIndex']=spdb['closeIndex']
    #计算回撤
    spdb['close_max'] = spdb['closeIndex'].expanding(min_periods=2).max()
    spdb['drawdown']=100*(spdb['closeIndex'] / spdb['close_max']-1)
    # maxdd = spdb['drawdown'].max()
    
    # 首尾对调
    spdb_w=spdb.iloc[::-1]
    #取周数据
    spdb_w=spdb_w.iloc[0:len(spdb_w):5]
    #首尾对调
    spdb_w=spdb_w.sort_index(ascending=1)
    spdb_w=spdb_w.reset_index(drop=True)
    
    spdb_w=spdb_w.rename(columns={'closeIndex':'closePrice'})
    spdb_w=spdb_w.rename(columns={'highestIndex':'highestPrice'})
    spdb_w=spdb_w.rename(columns={'lowestIndex':'lowestPrice'})

    dif,dea,macd = ta.MACD(spdb.closeIndex.values, fastperiod=20, slowperiod=40, signalperiod=9)
    spdb['KAMA_10']  = ta.KAMA(spdb.closeIndex.values, timeperiod=10)
    spdb['dif'] = dif
    spdb['dea'] = dea
    spdb['macd'] = 2*macd  # TA-Lib 中计算得到的macd=dif-dea，而国内一般计算macd=2*(dif-dea)，故此处乘以2，以求与国内习惯吻合
    spdb['tmp'] = [min(i, 0) for i in spdb.macd]
    
    if spdb.iloc[-1]['macd']>0 and spdb.iloc[-1]['macd']>1.03*spdb.iloc[-2]['macd']:
        MACD_result=u"涨"
    elif spdb.iloc[-1]['macd']<0 and spdb.iloc[-1]['macd']<0.97*spdb.iloc[-2]['macd']:
        MACD_result=u"跌"
    else:
        MACD_result=u"震荡"    
    
    spdb['diff'] = spdb['KAMA_10'].diff(periods=1)
    spdb['THRESHOLD'] = spdb['KAMA_10'].diff(periods=1).rolling(window=20,center=False).std()
    spdb['LLV4'] = spdb['KAMA_10'].rolling(window=4,center=False).min()
    spdb['HHV4'] = spdb['KAMA_10'].rolling(window=4,center=False).max()
    spdb['buy']=spdb['KAMA_10']-spdb['LLV4']
    spdb['sell']=spdb['HHV4']-spdb['KAMA_10']
    spdb["KAMA_10_buy"]=map(lambda x,y,z:z if (x>y) else np.nan,spdb["buy"],spdb["THRESHOLD"],spdb["KAMA_10"])
    spdb["KAMA_10_sell"]=map(lambda x,y,z:z if (x>y) else np.nan,spdb["sell"],spdb["THRESHOLD"],spdb["KAMA_10"])
    spdb["k"], spdb["d"],spdb["j"] =KDJ(spdb.closeIndex.values,spdb.highestIndex.values,spdb.lowestIndex.values, N=100,M1=3,M2=3)
    spdb=spdb.tail(250)
    spdb=spdb.reset_index(drop=True) 
    
    

    

    
    interval=40
    mylist=spdb.tradeDate.tolist()
    mylist=mylist[::-1]
    mylocs=range(len(mylist),0,0-interval)
    mylabels=mylist[0:len(mylist):interval]

    if np.isnan(spdb.iloc[-1]['KAMA_10_buy'])==False:
        KAMA_result=u"涨"
    elif np.isnan(spdb.iloc[-1]['KAMA_10_sell'])==False:
        KAMA_result=u"跌"
    else:
        KAMA_result=u"震荡"
        
    fig = plt.figure(figsize=(30,10))
    ax=plt.subplot(3, 2, 1)
    plt.plot(spdb.index,spdb['closeIndex'])
    plt.plot(spdb.index,spdb['KAMA_10'],color="b",linewidth=4)
    plt.plot(spdb.index,spdb['KAMA_10_buy'],color="r",linewidth=4)
    plt.plot(spdb.index,spdb['KAMA_10_sell'],color="g",linewidth=4)
    ax1 = ax.twinx()
    ax1.plot(spdb.index,spdb['drawdown'],color="gray",linewidth=2)
    plt.xticks(mylocs,mylabels)
    plt.title(spdb['secShortName'][0].decode("utf-8")+u" 日线"+u" 自适应均线(10): " +KAMA_result+u"   最大回撤："+str(round(spdb['drawdown'].min(),1))+u" 当前回撤："+str(spdb.iloc[-1]['drawdown'].round(1)) +"  "+KAMA_result,fontproperties=font,fontsize=12)

    ax=plt.subplot(3, 2, 3)
    plt.plot(spdb.index,spdb['dif'],color='c')
    plt.plot(spdb.index,spdb['dea'],color='y')
    plt.bar(spdb.index,spdb['macd'],color='r')
    plt.bar(spdb.index,spdb['tmp'],color='g')
    plt.xticks(mylocs,mylabels)
    plt.title(spdb['secShortName'][0].decode("utf-8")+u" 日线"+u" MACD(20,40,9)"+u" DIF:"+str(spdb.iloc[-1]['dif'].round(2)) +u" DEA:"+str(spdb.iloc[-1]['dea'].round(2))  +u" MACD:"+str(spdb.iloc[-1]['macd'].round(2))+"  "+MACD_result,fontproperties=font,fontsize=12)

    plt.subplot(3, 2, 5)
    plt.plot(spdb.index,spdb['k'])
    plt.plot(spdb.index,spdb['d'])
    plt.plot(spdb.index,spdb['j'])
    plt.xticks(mylocs,mylabels)
    plt.title(spdb['secShortName'][0].decode("utf-8")+u" 日线"+u" KDJ(100,3,3)"+u" K:"+str(spdb.iloc[-1]['k'].round(2)) +u" D:"+str(spdb.iloc[-1]['d'].round(2))  +u" J:"+str(spdb.iloc[-1]['j'].round(2)) ,fontproperties=font,fontsize=12)

    
    
    
    
    

    
    if myticker=='HSI':
        return


    dif_w,dea_w,macd_w = ta.MACD(spdb_w.closePrice.values, fastperiod=20, slowperiod=40, signalperiod=9)
    spdb_w['KAMA_10']  = ta.KAMA(spdb_w.closePrice.values, timeperiod=10)
    spdb_w['dif'] = dif_w
    spdb_w['dea'] = dea_w
    spdb_w['macd'] = 2*macd_w  # TA-Lib 中计算得到的macd=dif-dea，而国内一般计算macd=2*(dif-dea)，故此处乘以2，以求与国内习惯吻合
    spdb_w['tmp'] = [min(i, 0) for i in spdb_w.macd]
    
    if spdb_w.iloc[-1]['macd']>0 and spdb_w.iloc[-1]['macd']>1.03*spdb_w.iloc[-2]['macd']:
        MACD_w_result=u"涨"
    elif spdb_w.iloc[-1]['macd']<0 and spdb_w.iloc[-1]['macd']<0.97*spdb_w.iloc[-2]['macd']:
        MACD_w_result=u"跌"
    else:
        MACD_w_result=u"震荡"    
    
    spdb_w['diff'] = spdb_w['KAMA_10'].diff(periods=1)
    spdb_w['THRESHOLD'] = spdb_w['KAMA_10'].diff(periods=1).rolling(window=20,center=False).std()
    spdb_w['LLV4'] = spdb_w['KAMA_10'].rolling(window=4,center=False).min()
    spdb_w['HHV4'] = spdb_w['KAMA_10'].rolling(window=4,center=False).max()
    spdb_w['buy']=spdb_w['KAMA_10']-spdb_w['LLV4']
    spdb_w['sell']=spdb_w['HHV4']-spdb_w['KAMA_10']
    spdb_w["KAMA_10_buy"]=map(lambda x,y,z:z if (x>y) else np.nan,spdb_w["buy"],spdb_w["THRESHOLD"],spdb_w["KAMA_10"])
    spdb_w["KAMA_10_sell"]=map(lambda x,y,z:z if (x>y) else np.nan,spdb_w["sell"],spdb_w["THRESHOLD"],spdb_w["KAMA_10"])
    spdb_w["k"], spdb_w["d"],spdb_w["j"] =KDJ(spdb_w.closePrice.values,spdb_w.highestPrice.values,spdb_w.lowestPrice.values, N=100,M1=3,M2=3)
    spdb_w=spdb_w.tail(250)
    spdb_w=spdb_w.reset_index(drop=True) 

    interval_w=40
    mylist_w=spdb_w.endDate.tolist()
    mylist_w=mylist_w[::-1]
    mylocs_w=range(len(mylist_w),0,0-interval_w)
    mylabels_w=mylist_w[0:len(mylist_w):interval_w]

    if np.isnan(spdb_w.iloc[-1]['KAMA_10_buy'])==False:
        KAMA_result_w=u"涨"
    elif np.isnan(spdb_w.iloc[-1]['KAMA_10_sell'])==False:
        KAMA_result_w=u"跌"
    else:
        KAMA_result_w=u"震荡"

    ax=plt.subplot(3, 2, 2)
    plt.plot(spdb_w.index,spdb_w['closePrice'])
    plt.plot(spdb_w.index,spdb_w['KAMA_10'],color="b",linewidth=4)
    plt.plot(spdb_w.index,spdb_w['KAMA_10_buy'],color="r",linewidth=4)
    plt.plot(spdb_w.index,spdb_w['KAMA_10_sell'],color="g",linewidth=4)
    ax1 = ax.twinx()
    ax1.plot(spdb_w.index,spdb_w['drawdown'],color="gray",linewidth=2)
    plt.xticks(mylocs_w,mylabels_w)
    plt.title(spdb_w['secShortName'][0].decode("utf-8")+u" 周线"+u" 自适应均线(10): " +KAMA_result_w+u"   最大回撤："+str(round(spdb_w['drawdown'].min(),1))+u" 当前回撤："+str(spdb_w.iloc[-1]['drawdown'].round(1)) +"  "+KAMA_result_w ,fontproperties=font,fontsize=12)

    ax=plt.subplot(3, 2, 4)
    plt.plot(spdb_w.index,spdb_w['dif'],color='c')
    plt.plot(spdb_w.index,spdb_w['dea'],color='y')
    plt.bar(spdb_w.index,spdb_w['macd'],color='r')
    plt.bar(spdb_w.index,spdb_w['tmp'],color='g')
    plt.xticks(mylocs_w,mylabels_w)
    plt.title(spdb_w['secShortName'][0].decode("utf-8")+u" 周线"+u" MACD(20,40,9)"+u" DIF:"+str(spdb_w.iloc[-1]['dif'].round(2)) +u" DEA:"+str(spdb_w.iloc[-1]['dea'].round(2))  +u" MACD:"+str(spdb_w.iloc[-1]['macd'].round(2))+"  "+MACD_w_result,fontproperties=font,fontsize=12)

    plt.subplot(3, 2, 6)
    plt.plot(spdb_w.index,spdb_w['k'])
    plt.plot(spdb_w.index,spdb_w['d'])
    plt.plot(spdb_w.index,spdb_w['j'])
    plt.xticks(mylocs_w,mylabels_w)
    plt.title(spdb_w['secShortName'][0].decode("utf-8")+u" 周线"+u" KDJ(100,3,3)"+u" K:"+str(spdb_w.iloc[-1]['k'].round(2)) +u" D:"+str(spdb_w.iloc[-1]['d'].round(2))  +u" J:"+str(spdb_w.iloc[-1]['j'].round(2)) ,fontproperties=font,fontsize=12)   

def data_process(df,  is_stander = True):
    df_close = df.copy()
    ## 填充nan值，选择前一天收盘数据
    df_close.fillna(method= 'ffill', inplace= True)
    ## 还有nan值再填充
    df_close.fillna(method= 'bfill', inplace= True)
    ## 标准化
    if is_stander:
        model = StandardScaler()
        return pd.DataFrame(model.fit_transform(df_close), index= df_close.index, columns= df_close.columns)
    else:
        return df_close

def cluster_data(df_computer, df_label, n_cluster = 8):
    df_pro_close = df_computer.copy()
    df_close = df_label.copy()
    # print(df_pro_close.shape, df_close.shape)
    ## 聚类
    cluster_model = KMeans(n_clusters= n_cluster)
    cluster_model.fit(df_pro_close)
    ## label_data
    df_close['label'] = cluster_model.labels_
    dict_label = {k:v for k,v in zip(df_close.index, cluster_model.labels_)}
    return df_close, dict_label   
    
def plot_corr(df,size=10):
    df_corr = df.copy()
    fig, ax = plt.subplots(figsize = (size, size))
    cax = ax.matshow(df_corr, cmap = 'RdYlGn' )
    # plt.xticks(range(len(df_corr.columns)), df_corr.columns, rotation=90)
    # plt.yticks(range(len(df_corr.columns)), df_corr.columns)
    cbar = fig.colorbar(cax, ticks=[-1, 0, 1], aspect=40, shrink=.8)
    
    






def get_data_remove(ticker,reportDate=myreportDate):
    '''
    
    ''' 
    global data_inner
    
    #取管理公司名字
    company_name=DataAPI.FundGet(secID=u"",ticker=ticker,etfLof=u"",listStatusCd=u"",category=u"",idxID=u"",idxTicker=u"",operationMode=u"",beginDate="",endDate="",status="",field=u"ticker,secShortName,managementFullName",pandas="1")
    #取最近报告发布日期子基金列表
    data_temp=DataAPI.FundHoldingsGet(secID=u"",ticker=ticker,reportDate=reportDate,beginDate=u"",endDate=u"",secType="F",holdingTicker="",holdingSecID="",field=u"holdingTicker,holdingsecShortName,reportDate,ratioInNa,marketValue",pandas="1")
    if data_temp.empty:
        print(","),
        # print(company_name["secShortName"][0]+"子基金数据空缺")
        return data_temp
    data_temp=data_temp.sort_values('ratioInNa', ascending=False)
    #取前十重仓基金
    if myn==10:
        data_temp=data_temp.head(10)
    data_temp=data_temp.reset_index(drop=True)
    company_name_sub=DataAPI.FundGet(secID=u"",ticker=data_temp["holdingTicker"],etfLof=u"",listStatusCd=u"",category=u"",idxID=u"",idxTicker=u"",operationMode=u"",beginDate="",endDate="",status="A",field=u"ticker,managementFullName,",pandas="1")
    #去除重复
    company_name_sub = company_name_sub.groupby("ticker").max()
    company_name_sub['holdingTicker'] = company_name_sub.index
    company_name_sub.index = range(len(company_name_sub))
    if len(data_temp)!=len(company_name_sub):
        print("."),
      
    data_temp = pd.merge(data_temp, company_name_sub, on=['holdingTicker'], how='inner')
    #获取主要投资本公司基金的FOF
    data_temp_company=data_temp[data_temp["managementFullName"]==company_name["managementFullName"][0]]
    if len(data_temp_company)/len(data_temp)>=0.7:
        data_inner=pd.concat([data_inner, company_name], axis=0)
    #     print("投资本基金公司基金：")
        # print(company_name.to_html())    
        # print(data_inner.to_html()) 
    #获取非本FOF基金管理公司基金
    data_temp=data_temp[data_temp["managementFullName"]!=company_name["managementFullName"][0]]
    #增加购买数量为1
    data_temp["购买(只)"]=1
    data_temp.drop(['managementFullName'],axis=1,inplace=True)
    data_temp.drop(['ratioInNa'],axis=1,inplace=True)
    data_temp=data_temp.reset_index(drop=True)
    data_temp=data_temp.rename(columns={'marketValue':'购买(亿)'})
    data_temp["购买(亿)"]=map(lambda x:round(x/100000000,2), data_temp["购买(亿)"])
    return data_temp



def get_data(ticker,reportDate=myreportDate):
    '''
    
    ''' 
    #取最近报告发布日期子基金列表
    data_temp=DataAPI.FundHoldingsGet(secID=u"",ticker=ticker,reportDate=reportDate,beginDate=u"",endDate=u"",secType="F",holdingTicker="",holdingSecID="",field=u"holdingTicker,holdingsecShortName,reportDate,ratioInNa,marketValue",pandas="1")
    if data_temp.empty:
        # print(ticker+"FOF子基金数据空缺")
        return data_temp
    data_temp=data_temp.sort_values('ratioInNa', ascending=False)
    #取前十重仓基金
    if myn==10:
        data_temp=data_temp.head(10)
    data_temp=data_temp.reset_index(drop=True)

    return data_temp





def industry_fund(fund,weight, reportDate=myreportDate, mytitle="行业配置"):
    '''
    
    ''' 
    #取最近报告发布日期股票列表
    
    try:
        weight=weight.reset_index(drop=True)
    except:
        weight=weight
    else:
        weight=weight.reset_index(drop=True)   
    
    try:
        fund=fund.reset_index(drop=True)
    except:
        fund=fund
    else:
        fund=fund.reset_index(drop=True)     

    # fund=fund.reset_index(drop=True)
    # weight=weight.reset_index(drop=True)
    
    stock_fund_total = pd.DataFrame()
    for i in range(0, len(fund)):
        stock_fund=DataAPI.FundHoldingsGet(secID=u"",ticker=fund[i],reportDate=reportDate,beginDate=u"",endDate=u"",secType="E",holdingTicker="",holdingSecID="",field=u"holdingTicker,holdingsecShortName,reportDate,ratioInNa,marketValue",pandas="1")
        stock_fund=stock_fund.sort_values('ratioInNa', ascending=False)
        #取前十重仓基金
        if myn==10:
            stock_fund=stock_fund.head(10)
        stock_fund['ratioInNa'] = stock_fund['ratioInNa'].apply(lambda x: x*weight[i])
        stock_fund_total=pd.concat([stock_fund_total, stock_fund], axis=0)
        
    stock_fund_total=stock_fund_total.reset_index(drop=True)
    stock_fund_total = stock_fund_total.groupby("holdingTicker").sum()
    stock_fund_total['holdingTicker'] = stock_fund_total.index
    stock_fund_total.index = range(len(stock_fund_total))
    
    Eindustry=DataAPI.EquIndustryGet(secID=u"",ticker=stock_fund_total["holdingTicker"],industryVersionCD=u"010303",industry=u"",industryID=u"",industryID1=u"",industryID2=u"",industryID3=u"",intoDate=u"",equTypeID=u"",field=u"ticker,isNew,industryID1,industryName1",pandas="1")
    Eindustry=Eindustry.rename(columns={'ticker':'holdingTicker'})
    Eindustry=Eindustry[Eindustry["isNew"]==1]
    Eindustry.index = range(len(Eindustry))
    stock_fund_total = pd.merge(stock_fund_total, Eindustry, on=['holdingTicker'], how='outer')
    stock_fund_total = stock_fund_total.groupby("industryName1").sum()
    stock_fund_total['industryName1'] = stock_fund_total.index
    stock_fund_total.index = range(len(stock_fund_total))
    stock_fund_total=stock_fund_total.sort_values('ratioInNa', ascending=False)
    
    for j in range(0, len(stock_fund_total)):
        myname=stock_fund_total["industryName1"][j]
        if myname=="化工":
            stock_fund_total["industryName1"][j]=u"化工"
        elif myname=="电气设备":
            stock_fund_total["industryName1"][j]=u"电气设备"
        elif myname=="电子":
            stock_fund_total["industryName1"][j]=u"电子"   
        elif myname=="有色金属":
            stock_fund_total["industryName1"][j]=u"有色金属" 
        elif myname=="银行":
            stock_fund_total["industryName1"][j]=u"银行" 
        elif myname=="医药生物":
            stock_fund_total["industryName1"][j]=u"医药生物" 
        elif myname=="机械设备":
            stock_fund_total["industryName1"][j]=u"机械设备" 
        elif myname=="房地产":
            stock_fund_total["industryName1"][j]=u"房地产" 
        elif myname=="非银金融":
            stock_fund_total["industryName1"][j]=u"非银金融" 
        elif myname=="传媒":
            stock_fund_total["industryName1"][j]=u"传媒" 
        elif myname=="交通运输":
            stock_fund_total["industryName1"][j]=u"交通运输" 
        elif myname=="采掘":
            stock_fund_total["industryName1"][j]=u"采掘" 
        elif myname=="钢铁":
            stock_fund_total["industryName1"][j]=u"钢铁" 
        elif myname=="食品饮料":
            stock_fund_total["industryName1"][j]=u"食品饮料" 
        elif myname=="公用事业":
            stock_fund_total["industryName1"][j]=u"公用事业" 
        elif myname=="汽车":
            stock_fund_total["industryName1"][j]=u"汽车" 
        elif myname=="纺织服装":
            stock_fund_total["industryName1"][j]=u"纺织服装" 
        elif myname=="建筑装饰":
            stock_fund_total["industryName1"][j]=u"建筑装饰" 
        elif myname=="国防军工":
            stock_fund_total["industryName1"][j]=u"国防军工" 
        elif myname=="家用电器":
            stock_fund_total["industryName1"][j]=u"家用电器" 
        elif myname=="综合":
            stock_fund_total["industryName1"][j]=u"综合" 
        elif myname=="建筑材料":
            stock_fund_total["industryName1"][j]=u"建筑材料" 
        elif myname=="休闲服务":
            stock_fund_total["industryName1"][j]=u"休闲服务" 
        elif myname=="计算机":
            stock_fund_total["industryName1"][j]=u"计算机" 
        elif myname=="农林牧渔":
            stock_fund_total["industryName1"][j]=u"农林牧渔" 
        elif myname=="轻工制造":
            stock_fund_total["industryName1"][j]=u"轻工制造" 
        elif myname=="通信":
            stock_fund_total["industryName1"][j]=u"通信" 
        elif myname=="商业贸易":
            stock_fund_total["industryName1"][j]=u"商业贸易"    
        
    plt.figure(figsize=(10, 10))
    plt.axis('equal')
    plt.pie(x=stock_fund_total["ratioInNa"], autopct='%1.1f%%')
    plt.legend(stock_fund_total.head(10)["industryName1"], prop=small_font)
    plt.title(mytitle,fontproperties=font)
    
    stock_fund_total.drop(['isNew'],axis=1,inplace=True)
    stock_fund_total.drop(['marketValue'],axis=1,inplace=True)
    stock_fund_total.reset_index(drop=True,inplace=True)

    # return stock_fund_total
    return


def industry_FOF(FOF,weight, reportDate=myreportDate,mytitle="行业配置"):
    
    stock_FOF_total = pd.DataFrame()
    for i in range(0, len(FOF)):
        stock_FOF=get_data(FOF[i],reportDate)
        stock_FOF=stock_FOF.sort_values('ratioInNa', ascending=False)
        #取前十重仓基金
        if myn==10:
            stock_FOF=stock_FOF.head(10)
        stock_FOF['ratioInNa'] = stock_FOF['ratioInNa'].apply(lambda x: x*weight[i])
        stock_FOF_total=pd.concat([stock_FOF_total, stock_FOF], axis=0)
    
    stock_FOF_total=stock_FOF_total.reset_index(drop=True)
    stock_FOF_total = stock_FOF_total.groupby("holdingTicker").sum()
    stock_FOF_total['holdingTicker'] = stock_FOF_total.index
    stock_FOF_total.index = range(len(stock_FOF_total))
    bbbbb=industry_fund(stock_FOF_total['holdingTicker'],stock_FOF_total['ratioInNa'],reportDate,mytitle)

    return bbbbb




def my_sort(data_temp,*value):
#排序
    j=0
    data_temp["myresult"]=0
    for i in value:
        if i[1]==True:
            data_temp[i[0]+"1"]=-data_temp[i[0]]
        else:
            data_temp[i[0]+"1"]=data_temp[i[0]]
        data_temp[i[0]+"1"]=(data_temp[i[0]+"1"]-data_temp[i[0]+"1"].mean())/data_temp[i[0]+"1"].std()
        data_temp["myresult"]=data_temp["myresult"]+i[2]*data_temp[i[0]+"1"]
        data_temp.drop([i[0]+"1"],axis=1,inplace=True)
    data_temp=data_temp.sort_values('myresult', ascending=False)
    data_temp=data_temp.reset_index(drop=True) 
    data_temp.drop(["myresult"],axis=1,inplace=True)
    # data_temp['holdingTicker'] = data_temp.index
    data_temp.index = range(len(data_temp))
 
    return data_temp




def show_result(data_temp,fund_name,n=number_show):
    #显示结果

    print("")
    print(fund_name+"总数: "+str(len(data_temp))+", "+"以下只显示前"+str(min(n,len(data_temp)))+"只基金收益")
    
    # if scale==1:
    #     data_temp=my_sort(data_temp, ["一月",False,1.1], ["三月",False,1],["六月",False,0.9],["一年",False,0.8],["内部持有",False,1],["机构持有",False,1],["CALMAR",False,1],["夏普比率",False,1],["规模(亿)",True,1],["份额(亿)",True,1],["最大回撤",True,1])
    # else:
    #     data_temp=my_sort(data_temp, ["一月",False,1.1], ["三月",False,1],["六月",False,0.9],["一年",False,0.8],["内部持有",False,1],["机构持有",False,1],["CALMAR",False,1],["夏普比率",False,1],["最大回撤",True,1])

    print("一月收益："),
    print(data_temp.head(n)["一月"].median())
    print("三月收益："),
    print(data_temp.head(n)["三月"].median())
    print("六月收益："),
    print(data_temp.head(n)["六月"].median())
    print("一年收益："),
    print(data_temp.head(n)["一年"].median())    

        
    print(data_temp.head(n).to_html())
    return


def period_year(x):

    if x is np.nan:
        return np.nan

    else:
        before = datetime.strptime(x, '%Y-%m-%d').date()
        now = datetime.strptime(get_recent_date(0), '%Y-%m-%d').date()
        Days = (now - before).days
        return round(Days/365.0, 1)        




def add_manager(data_temp):
    #基金经理，业绩基准

    a=float(len(data_temp)/10.0)
    print ""
    print "获取基金经理数据......",

    manager_name=DataAPI.FundManagerGet(secID=u"",ticker=data_temp["holdingTicker"],personID=u"",name=u"",partyID="",field=u"",pandas="1")
    manager_name=manager_name[manager_name['isIncumbent']==1]
    manager_name=manager_name[manager_name['position']=="FM"]
    # print(manager_name.to_html())  
    manager_name=DataAPI.FundManagerInfoGet(personID=manager_name["personID"],name=u"",genderCD=u"",field=u"",pandas="1")
    manager_name["period"]=""
    data_temp["period"]=""
    manager_name["period"]=map(period_year, manager_name["practiceDate"]) 
    k=0
    for j in range(0, len(data_temp)):
        myname=data_temp["managerName"][j]

        if "、" in myname:
            myname1=myname[:myname.index("、")]
            myname2=myname[myname.index("、")+3:]
            index = manager_name.loc[(manager_name.name==myname1)].index.tolist()
        else:
            index = manager_name.loc[(manager_name.name==myname)].index.tolist()
        
        if index:
            data_temp["period"][j]=manager_name["period"][index[0]]
    return data_temp

def add_category(data_temp):


    #获取基金类型
    aaa=DataAPI.FundGet(secID=u"",ticker=data_temp["holdingTicker"],etfLof=u"",listStatusCd=u"",category=u"",idxID=u"",idxTicker=u"",operationMode=u"",beginDate="",endDate="",status="A",field=u"ticker,secShortName,managerName,perfBenchmark,category,etfLof",pandas="1")
    aaa = aaa.groupby("ticker").max()
    aaa['holdingTicker'] = aaa.index
    aaa.index = range(len(aaa))
    data_temp = pd.merge(data_temp, aaa, on=['holdingTicker'], how='inner')

    data_temp["category"]=map(lambda x,y:"ETF基金" if x=="ETF" else y,data_temp["etfLof"],data_temp["category"])

    data_temp["category"].replace("E","股票基金",inplace=True)
    data_temp["category"].replace("H","混合基金",inplace=True)
    data_temp["category"].replace("B","债券基金",inplace=True)
    data_temp["category"].replace("SB","短期理财债券基金",inplace=True)
    data_temp["category"].replace("M","货币基金",inplace=True)
    data_temp["category"].replace("O","其它基金",inplace=True)
    data_temp.drop(['etfLof'],axis=1,inplace=True)

    return data_temp


def add_netasset(data_temp, reportDate=myreportDate, show=True):
    #基金规模,股票占比

    a=float(len(data_temp)/10.0)
    if show:
        print ""
        print "获取基金规模数据......",

    data_temp=data_temp.reset_index(drop=True)
    net_asset=DataAPI.FundAssetsGet(secID=u"",ticker=data_temp["holdingTicker"],reportDate=reportDate,updateTime=u"",beginDate=u"",endDate=u"",field=u"ticker,reportDate,equityRatioInTa,netAsset",pandas="1")
    #取最近报告期
    # net_asset=net_asset.sort('reportDate', ascending=False).groupby('ticker', as_index=False).first()
    net_asset=net_asset.rename(columns={'ticker':'holdingTicker'})

    data_temp = pd.merge(data_temp, net_asset, on=['holdingTicker'], how='outer')

    data_temp["规模(亿)"]=map(lambda x:round(x/100000000,1), data_temp["netAsset"])
    data_temp["股票占比"]=map(lambda x:round(x,1), data_temp["equityRatioInTa"])
    data_temp.drop(['netAsset'],axis=1,inplace=True)
    data_temp.drop(['equityRatioInTa'],axis=1,inplace=True)
    data_temp["股票占比"].fillna(0,inplace=True)
  
    return data_temp

def add_holder(data_temp):
    #持有人结构

    print ""
    print "获取基金持有人数据......",


    fund_holder=DataAPI.FundHolderInfoGet(ticker=data_temp["holdingTicker"],secID=u"",beginDate=u"",endDate=u"",infoSourceCD=u"",field=u"ticker,reportDate,holderNumber,avgHoldShares,instHoldRatio,pracHoldRatio",pandas="1")
    fund_holder["份额(亿)"]=fund_holder["holderNumber"]*fund_holder["avgHoldShares"]/100000000
    
    fund_holder=fund_holder.sort('reportDate', ascending=False).groupby('ticker', as_index=False).first()
    fund_holder.drop(['reportDate'],axis=1,inplace=True)
    fund_holder.drop(['holderNumber'],axis=1,inplace=True)
    fund_holder.drop(['avgHoldShares'],axis=1,inplace=True)

    fund_holder=fund_holder.rename(columns={'ticker':'holdingTicker'}) 

    data_temp = pd.merge(data_temp, fund_holder, on=['holdingTicker'], how='outer')
    data_temp=data_temp.rename(columns={'instHoldRatio':'机构持有'}) 
    data_temp=data_temp.rename(columns={'pracHoldRatio':'内部持有'}) 
    data_temp["份额(亿)"]=map(lambda x:round(x,1), data_temp["份额(亿)"]) 
    return data_temp


def add_fundnav(data_temp):
    #基金动量

    print ""
    print "获取基金趋势数据......",
    
    nav_0=DataAPI.FundNavGet(secID=u"",ticker=data_temp["holdingTicker"],dataDate="",beginDate=get_recent_date(260),endDate="",partyID="",partyShortName="",field=u"ticker,endDate,ADJUST_NAV",pandas="1")
    nav_0=nav_0.groupby("ticker").apply(lambda x:x.sort_values(by = 'endDate',ascending=True).iloc[-1,:])
    nav_0=nav_0.rename(columns={'ticker':'holdingTicker', 'ADJUST_NAV':'ADJUST_NAV_0'})  
    nav_0.drop(['endDate'],axis=1,inplace=True)
    nav_30=DataAPI.FundNavGet(secID=u"",ticker=data_temp["holdingTicker"],dataDate=get_recent_date(20),beginDate="",endDate="",partyID="",partyShortName="",field=u"ticker,ADJUST_NAV",pandas="1")
    nav_30=nav_30.rename(columns={'ticker':'holdingTicker', 'ADJUST_NAV':'ADJUST_NAV_30'}) 
    nav_90=DataAPI.FundNavGet(secID=u"",ticker=data_temp["holdingTicker"],dataDate=get_recent_date(61),beginDate="",endDate="",partyID="",partyShortName="",field=u"ticker,ADJUST_NAV",pandas="1")
    nav_90=nav_90.rename(columns={'ticker':'holdingTicker', 'ADJUST_NAV':'ADJUST_NAV_90'}) 
    nav_180=DataAPI.FundNavGet(secID=u"",ticker=data_temp["holdingTicker"],dataDate=get_recent_date(122),beginDate="",endDate="",partyID="",partyShortName="",field=u"ticker,ADJUST_NAV",pandas="1")
    nav_180=nav_180.rename(columns={'ticker':'holdingTicker', 'ADJUST_NAV':'ADJUST_NAV_180'}) 
    nav_360=DataAPI.FundNavGet(secID=u"",ticker=data_temp["holdingTicker"],dataDate=get_recent_date(245),beginDate="",endDate="",partyID="",partyShortName="",field=u"ticker,ADJUST_NAV",pandas="1")
    nav_360=nav_360.rename(columns={'ticker':'holdingTicker', 'ADJUST_NAV':'ADJUST_NAV_360'}) 
    data_temp = pd.merge(data_temp, nav_0, on=['holdingTicker'], how='outer')
    data_temp = pd.merge(data_temp, nav_30, on=['holdingTicker'], how='outer')
    data_temp = pd.merge(data_temp, nav_90, on=['holdingTicker'], how='outer')
    data_temp = pd.merge(data_temp, nav_180, on=['holdingTicker'], how='outer')
    data_temp = pd.merge(data_temp, nav_360, on=['holdingTicker'], how='outer')
    data_temp["一月"]=100*(data_temp['ADJUST_NAV_0']-data_temp['ADJUST_NAV_30'])/data_temp['ADJUST_NAV_30']
    data_temp["三月"]=100*(data_temp['ADJUST_NAV_0']-data_temp['ADJUST_NAV_90'])/data_temp['ADJUST_NAV_90']
    data_temp["六月"]=100*(data_temp['ADJUST_NAV_0']-data_temp['ADJUST_NAV_180'])/data_temp['ADJUST_NAV_180']
    data_temp["一年"]=100*(data_temp['ADJUST_NAV_0']-data_temp['ADJUST_NAV_360'])/data_temp['ADJUST_NAV_360']

    data_temp.drop(['ADJUST_NAV_0'],axis=1,inplace=True)
    data_temp.drop(['ADJUST_NAV_30'],axis=1,inplace=True)
    data_temp.drop(['ADJUST_NAV_90'],axis=1,inplace=True)
    data_temp.drop(['ADJUST_NAV_180'],axis=1,inplace=True)
    data_temp.drop(['ADJUST_NAV_360'],axis=1,inplace=True)
    
    # data_temp["短期"]=map(lambda x:"↗" if x>5 else ("↘" if x<-5 else "→" ) ,data_temp["短期"])
    # data_temp["中期"]=map(lambda x:"↗" if x>5 else ("↘" if x<-5 else "→" ) ,data_temp["中期"])
    # data_temp["长期"]=map(lambda x:"↗" if x>5 else ("↘" if x<-5 else "→" ) ,data_temp["长期"])
    
    # print(len(data_temp))
    # print(data_temp.to_html())    

    return data_temp


def add_performance(data_temp,window=7):
    #基金年化收益,最大回撤,夏普比率

    print ""
    print "获取基金性能数据......",
    
    # fund_performance=DataAPI.FundPerfIndicGet(dataDate=get_recent_date(0),secID=u"",ticker=data_temp["holdingTicker"],window=str(window),beginDate=u"",endDate=u"",field=u"window,endDate,ticker,annTotReturn,sharpeRatio,maxDD,realizedAlpha,CALMAR",pandas="1")
    fund_performance=DataAPI.FundPerfIndicGet(dataDate=u"",secID=u"",ticker=data_temp["holdingTicker"],window=str(window),beginDate=u"",endDate=u"",field=u"window,endDate,ticker,annTotReturn,sharpeRatio,maxDD,realizedAlpha,CALMAR",pandas="1")
    fund_performance=fund_performance.groupby("ticker").apply(lambda x:x.sort_values(by = 'endDate',ascending=True).iloc[-1,:])
    fund_performance=fund_performance.rename(columns={'ticker':'holdingTicker'})  

    data_temp = pd.merge(data_temp, fund_performance, on=['holdingTicker'], how='outer')
 
    data_temp["年化收益"]=map(lambda x:round(100*x,1), data_temp["annTotReturn"])
    data_temp["最大回撤"]=map(lambda x:round(100*x,1), data_temp["maxDD"])
    data_temp["夏普比率"]=map(lambda x:round(x,2), data_temp["sharpeRatio"])
    data_temp["ALPHA"]=map(lambda x:round(x,3), data_temp["realizedAlpha"])
    data_temp.drop(['annTotReturn'],axis=1,inplace=True)
    data_temp.drop(['maxDD'],axis=1,inplace=True)
    data_temp.drop(['sharpeRatio'],axis=1,inplace=True)
    data_temp.drop(['realizedAlpha'],axis=1,inplace=True)
    data_temp.drop(['window'],axis=1,inplace=True)
    # data_temp.drop(['endDate'],axis=1,inplace=True)
   
    return data_temp


def get_industry_staff(x):
    
    fund_holdings_stock_factor_sum = x.groupby("industryName1").sum()
    fund_holdings_stock_factor_sum=fund_holdings_stock_factor_sum.sort_values('ratioInNa_holding', ascending=False)
    fund_holdings_stock_factor_sum["industy"]=fund_holdings_stock_factor_sum.index
    # print(fund_holdings_stock_factor_sum.to_html())
    fund_holdings_stock_factor_sum["ratioInNa_holding"]=100*fund_holdings_stock_factor_sum["ratioInNa_holding"]/fund_holdings_stock_factor_sum["ratioInNa_holding"].sum()
    fund_holdings_stock_factor_sum["ratioInNa_holding"]=fund_holdings_stock_factor_sum["ratioInNa_holding"].astype(int)
    # print(fund_holdings_stock_factor_sum.to_html())
    myresult="|"
    for i in range(0, min(5,len(fund_holdings_stock_factor_sum))):
        myresult=myresult+fund_holdings_stock_factor_sum["industy"][i]+str(fund_holdings_stock_factor_sum["ratioInNa_holding"][i])+"|"
    # print(myresult)
    return myresult


def add_fundholdings(data_temp, reportDate=myreportDate):
    #基金十大

    stock_factor=DataAPI.MktStockFactorsOneDayGet(tradeDate=get_recent_date(1),secID=u"",ticker=u"",field=u"tradeDate,ticker,LCAP,PE,PB,ROE",pandas="1")
    stock_factor['LCAP']=map(lambda x:round(np.e**x/100000000,1), stock_factor['LCAP'])
    stock_factor.rename(columns={'ticker':'holdingTicker'},inplace=True)
    Eindustry=DataAPI.EquIndustryGet(secID=u"",ticker=stock_factor["holdingTicker"],industryVersionCD=u"010321",industry=u"",industryID=u"",industryID1=u"",industryID2=u"",industryID3=u"",intoDate=u"",equTypeID=u"",field=u"ticker,isNew,industryID1,industryName1",pandas="1")
    Eindustry=Eindustry.rename(columns={'ticker':'holdingTicker'})
    Eindustry=Eindustry[Eindustry["isNew"]==1]
    Eindustry.index = range(len(Eindustry))
    stock_factor = pd.merge(stock_factor, Eindustry, on=['holdingTicker'], how='outer')
    print ""
    print "获取基金股票持仓数据......",
    fund_holdings=DataAPI.FundHoldingsGet(secID=u"",ticker=data_temp["holdingTicker"],reportDate=reportDate,beginDate=u"",endDate=u"",secType="E",holdingTicker="",holdingSecID="",field=u"ticker,SecShortName,reportDate,holdingTicker,holdingsecShortName,ratioInNa",pandas="1")

    if fund_holdings["reportDate"].max()!=fund_holdings["reportDate"].min():
        print("error，取到两个报告期")
        print(fund_holdings[fund_holdings['reportDate']==fund_holdings['reportDate'].max()])
    fund_holdings.rename(columns={'ratioInNa':'ratioInNa_holding'},inplace=True)
    fund_holdings_stock_factor = pd.merge(stock_factor, fund_holdings, on=['holdingTicker'], how='inner')
    ticker_industry=fund_holdings_stock_factor.groupby("ticker").apply(get_industry_staff)
    # print(ticker_industry)
    ticker_industry=pd.DataFrame({"holdingTicker":ticker_industry.index,"前五行业":ticker_industry.values})
    #子基金股票中位数    
    fund_holdings_stock_factor_median = fund_holdings_stock_factor.groupby("ticker").median()
    fund_holdings_stock_factor_median['ticker'] = fund_holdings_stock_factor_median.index
    fund_holdings_stock_factor_median.index = range(len(fund_holdings_stock_factor_median))
    #子基金前十股票仓位占比求和
    fund_holdings_stock_factor=fund_holdings_stock_factor.sort_values('ratioInNa_holding', ascending=False)
    fund_holdings_stock_factor= fund_holdings_stock_factor.groupby("ticker").head(10)
    #------------------------------------------------------------------
    fund_holdings_stock_factor_sum = fund_holdings_stock_factor.groupby("ticker").sum()
    fund_holdings_stock_factor_sum['ticker'] = fund_holdings_stock_factor_sum.index
    fund_holdings_stock_factor_sum.index = range(len(fund_holdings_stock_factor_sum))
    fund_holdings_stock_factor_median=fund_holdings_stock_factor_median.rename(columns={'ticker':'holdingTicker'}) 
    fund_holdings_stock_factor_sum=fund_holdings_stock_factor_sum.rename(columns={'ticker':'holdingTicker'}) 
    fund_holdings_stock_factor_median_merge = pd.merge(data_temp, fund_holdings_stock_factor_median, on=['holdingTicker'], how='outer')
    fund_holdings_stock_factor_sum_merge = pd.merge(data_temp, fund_holdings_stock_factor_sum, on=['holdingTicker'], how='outer')
    
    
    
    if fund_holdings_stock_factor_median_merge["holdingTicker"].equals(fund_holdings_stock_factor_sum_merge["holdingTicker"])==False:
        print("基金列表错误。。。。。。")
        print(fund_holdings_stock_factor_sum_merge.to_html())
    fund_holdings_stock_factor_median_merge["ratioInNa_holding"]=fund_holdings_stock_factor_sum_merge["ratioInNa_holding"]
    fund_holdings_stock_factor_median_merge.drop(["PB"],axis=1,inplace=True)
    fund_holdings_stock_factor_median_merge.drop(["ROE"],axis=1,inplace=True)
    fund_holdings_stock_factor_median_merge.rename(columns={'LCAP':'股票市值'},inplace=True)
    fund_holdings_stock_factor_median_merge.rename(columns={'ratioInNa_holding':'前十股票占比'},inplace=True)
    fund_holdings_stock_factor_median_merge["PE"]=map(lambda x:round(x,1), fund_holdings_stock_factor_median_merge["PE"])     
    #取转债
    end_date = (datetime.strptime(myreportDate,'%Y-%m-%d') + timedelta(days=30)).strftime('%Y%m%d')
    fund_holdings=DataAPI.FundBondTypeGet(secID=u"",ticker=data_temp["holdingTicker"],beginDate=myreportDate,endDate=end_date,bondTypeCD=u"",field=u"ticker,bondType,valueInNA",pandas="1")
    fund_holdings=fund_holdings.rename(columns={'ticker':'holdingTicker'}) 
    fund_holdings=fund_holdings[fund_holdings["bondType"]=="可转换债券"]
    fund_holdings=fund_holdings.rename(columns={'valueInNA':'转债'}) 
    fund_holdings.drop(["bondType"],axis=1,inplace=True)
    fund_holdings_stock_factor_median_merge = pd.merge(fund_holdings_stock_factor_median_merge, fund_holdings, on=['holdingTicker'], how='outer')
    fund_holdings_stock_factor_median_merge["转债"].fillna(0,inplace=True)
    fund_holdings_stock_factor_median_merge["category"]=map(lambda x,y:"转债基金" if (x=="债券基金") and (y>=60) else x,fund_holdings_stock_factor_median_merge["category"],fund_holdings_stock_factor_median_merge["转债"])
    fund_holdings_stock_factor_median_merge = pd.merge(fund_holdings_stock_factor_median_merge, ticker_industry, on=['holdingTicker'], how='outer')
    fund_holdings_stock_factor_median_merge.drop(["isNew"],axis=1,inplace=True)
    return fund_holdings_stock_factor_median_merge


def add_fund_all(data_temp):
    #基金年化收益,最大回撤,夏普比率......
    
    #获取基金类型
    data_temp=add_category(data_temp)
    #增加基金经理年限
    data_temp=add_manager(data_temp)
    # # 增加基金规模，股票占比
    data_temp=add_netasset(data_temp)
    # # 增加前十大持仓占比
    data_temp=add_fundholdings(data_temp)
    # # 增加基金性能数据
    #三年
    # data_temp=add_performance(data_temp,7)
    #一年
    data_temp=add_performance(data_temp,5)
    # # 增加基金动量数据
    data_temp=add_fundnav(data_temp)
    # #增加基金持有人信息
    data_temp=add_holder(data_temp)
    data_temp=data_temp.round(2)
    return data_temp

def fund_classification(data_temp):
    #基金分类......

    data_temp.reset_index(drop=True,inplace=True)
    myticker=data_temp['holdingTicker']
    mydata_temp = pd.DataFrame()
    for i in range(0, len(myticker)):
        # df_close_1=DataAPI.FundNavGet(secID=u"",ticker=myticker[i],dataDate=u"",beginDate=get_recent_date(245),endDate=get_recent_date(0),partyID="",partyShortName="",field=u"endDate,ADJUST_NAV",pandas="1").set_index('endDate')
        df_close_1=DataAPI.FundNavGet(secID=u"",ticker=myticker[i],dataDate=u"",beginDate=get_recent_date(60),endDate=get_recent_date(0),partyID="",partyShortName="",field=u"endDate,ADJUST_NAV",pandas="1").set_index('endDate')
        df_close_1.rename(columns={'ADJUST_NAV':myticker[i]}, inplace = True)
        mydata_temp=pd.concat([mydata_temp, df_close_1], axis=1)

    mydata_temp=mydata_temp.T

    df_close=mydata_temp

    df_pro_close = data_process(df_close.T)

    df_close_label, dict_label  = cluster_data(df_pro_close.T, df_close, n_cluster=  n_cluster)

    mydict_label=pd.Series(dict_label)

    for i in range(0, n_cluster):
        myindex=mydict_label[mydict_label.values==i].index
        mydata_temp=data_temp[data_temp["holdingTicker"].isin(myindex)]
        print("第"+str(i)+"类基金:"),
        print(len(mydata_temp))
        print("")
        print("一月："),
        print(mydata_temp.head(50)["一月"].median())
        print("三月："),
        print(mydata_temp.head(50)["三月"].median())
        print("六月："),
        print(mydata_temp.head(50)["六月"].median())
        print("一年："),
        print(mydata_temp.head(50)["一年"].median())
        print(mydata_temp.head(50).to_html())

        weight=[1]*len(mydata_temp)
        # industry_fund(mydata_temp["ticker"],mydata_temp["购买(亿)"],myreportDate,u"行业配置:")
        industry_fund(mydata_temp["holdingTicker"],weight,myreportDate,u"行业配置"+str(i))
        
    print('聚类前相关性集群')
    plot_corr(df_close.T.corr() , size = 16)
    plt.show()
    print('聚类后相关性集群')
    plot_corr(df_close.T[pd.Series(dict_label).sort_values().index].corr() , size = 16)
    plt.show()

    # return data_temp
    return



#-------------------------生成FOF基金列表-------------------------------------------

#程序从此处开始
starttime = datetime.now()

#三个月之前的日期
start_date = (datetime.strptime(get_recent_date(0),'%Y-%m-%d') + timedelta(days=-90)).strftime('%Y%m%d')
#所有三个月之前成立的存续基金，去除成立不满三个月的基金
aaa=DataAPI.FundGet(secID=u"",ticker=u"",etfLof=u"",listStatusCd=u"",category=u"",idxID=u"",idxTicker=u"",operationMode=u"",beginDate="",endDate=start_date,status="A",field=u"secID,ticker,secShortName,establishDate,managerName,perfBenchmark,isFof,isQdii",pandas="1")
aaa["holdingTicker"]=aaa["ticker"]

#去除非FOF基金
bbb=aaa[~aaa['isFof'].isin([0])]
#去除QDII类基金
bbb=bbb[~bbb['isQdii'].isin([1])]
#去除C类基金
bbb=bbb[~ bbb['secShortName'].str.contains('C')]
bbb=bbb.reset_index(drop=True) 


# print(get_recent_date(245))

#生成FOF列表
myset=bbb["ticker"]

#-------------------------生成FOF基金列表-------------------------------------------


#------------------------生成FOF子基金性能-----------------------------


#获取上一报告期
index_month=myreportDate.find('-')+1
month_reportdate=myreportDate[index_month:index_month+2]
year_reportdate=myreportDate[:index_month-1]
if month_reportdate=="03":
    month_reportdate_before="12"
    year_reportdate_before=str(int(year_reportdate)-1)
    day_reportdate_before="31"
elif month_reportdate=="06":
    month_reportdate_before="03"
    year_reportdate_before=year_reportdate
    day_reportdate_before="31"
elif month_reportdate=="09":
    month_reportdate_before="06"
    year_reportdate_before=year_reportdate
    day_reportdate_before="30"
elif month_reportdate=="12":
    month_reportdate_before="09"
    year_reportdate_before=year_reportdate
    day_reportdate_before="30"
myreportDate_before=year_reportdate_before+"-"+month_reportdate_before+"-"+day_reportdate_before
#----------------------------------------------------------------

data_temp = pd.DataFrame()


print("获取FOF基金数据，进度(到10结束):"),
a=float(len(myset)/10.0)
k=0
for i in range(0, len(myset)):
# for i in range(0, 10):
    if k!=int(i/a):
        k=int(i/a)
        print(k),

    data_temp=pd.concat([data_temp, get_data_remove(myset[i],myreportDate)], axis=0)
    # data_temp_before=pd.concat([data_temp_before, get_data_remove(myset[i],myreportDate_before)], axis=0)
    
print("")
data_temp = data_temp.groupby("holdingTicker").sum()
data_temp['holdingTicker'] = data_temp.index
data_temp.index = range(len(data_temp))
data_temp=add_fund_all(data_temp)

# print(len(data_temp))

#提取对冲绝对收益基金
fund_hedge=data_temp[data_temp['secShortName'].str.contains('对冲')]
fund_hedge=pd.concat([fund_hedge, data_temp[data_temp['secShortName'].str.contains('绝对收益')]], axis=0)
print("")
print("")
print("对冲绝对收益基金")
print(fund_hedge.to_html())

# #去掉对冲混合-A
data_temp=data_temp[~data_temp['secShortName'].str.contains('对冲')]
data_temp=data_temp[~data_temp['secShortName'].str.contains('绝对收益')]
data_temp.reset_index(drop=True,inplace=True)



print("")
print("主要投资本基金公司子基金的FOF")
print(data_inner.to_html())

#基金经理工作年限大于4年
# data_temp=data_temp[data_temp['period']>4]    

data_temp["股+转债"]=data_temp["股票占比"]+data_temp["转债"]/2

#按照FOF购买数量，购买市值, Alpha,最大回撤综合排序    
data_temp=my_sort(data_temp, ["一月",False,1.1], ["三月",False,1],["六月",False,0.9],["一年",False,0.8] ,["购买(只)",False,1],["购买(亿)",False,1],["最大回撤",True,1],["CALMAR",False,1],["夏普比率",False,1])

print("")

print("按照FOF购买数量，购买市值, 年化收益,最大回撤综合排序")


data_temp1=data_temp[(data_temp['股+转债']<=10) & (data_temp['category']!="转债基金")]
# data_temp1=data_temp1[data_temp1['CALMAR']>=2]
show_result(data_temp1,"纯债",30)


data_temp1=data_temp[(data_temp['股+转债']>10) & (data_temp['股+转债']<=40) & (data_temp['category']!="转债基金")]
# data_temp1=data_temp1[data_temp1['CALMAR']>=2]
show_result(data_temp1,"固收+",30)


data_temp1=data_temp[data_temp['category']=="转债基金"]
show_result(data_temp1,"转债基金",len(data_temp1))



# print("按照FOF购买数量，购买市值, 年化收益,最大回撤，规模综合排序")
data_temp=my_sort(data_temp, ["一月",False,1.1], ["三月",False,1],["六月",False,0.9],["一年",False,0.8],["内部持有",False,1],["购买(只)",False,1],["购买(亿)",False,1],["CALMAR",False,1],["夏普比率",False,1],["规模(亿)",True,1],["份额(亿)",True,1],["最大回撤",True,1])


data_temp31=data_temp[(data_temp['股票占比']>40) & (data_temp['股票占比']<60)]
show_result(data_temp31,"均衡混合")

print("沪港深基金：")
data_temp31=data_temp[data_temp['secShortName'].str.contains('沪港深')]
show_result(data_temp31,"沪港深")
# for i in data_temp31.iloc[-10:-1]["holdingTicker"]:
#     analysis_stock(i)

print("指数增强基金：")
data_temp31=data_temp[data_temp['secShortName'].str.contains('指数')]
data_temp31=data_temp31[data_temp31['secShortName'].str.contains('增强')]
show_result(data_temp31,"指数增强")
    
    
#选出股票占比大于60,基金规模大于5
data_temp31=data_temp[data_temp['股票占比']>60]
data_temp31=data_temp31[data_temp31['规模(亿)']>5]
data_temp31=data_temp31[data_temp31["category"]!="ETF基金"]

data_temp31=data_temp31.dropna(axis=0,how="any")

data_temp32=my_sort(data_temp31, ["购买(只)",False,1],["购买(亿)",False,1])
show_result(data_temp32,"FOF热门",30)
weight=[1]*len(data_temp32.head(30))
industry_fund(data_temp32.head(30)["holdingTicker"],weight,myreportDate,u"当季FOF热门前30基金行业配置:")
# for i in data_temp32.iloc[0:10]["holdingTicker"]:
#     analysis_stock(i)

print("股票占比大于60%，规模大于5亿的基金综合排序: 年化收益0.2+最大回撤0.1+基金规模0.1+FOF购买数量0.1+FOF购买规模0.1+基金从业人员持有比例*0.1")
show_result(data_temp31,"当季综合股混前30",30)
weight=[1]*len(data_temp31.head(30))
industry_fund(data_temp31.head(30)["holdingTicker"],weight,myreportDate,u"当季综合股混前30基金行业配置:")
# for i in data_temp31.iloc[0:10]["holdingTicker"]:
#     analysis_stock(i)


show_result(data_temp31.tail(30),"当季综合股混最后30",30)
weight=[1]*len(data_temp31.tail(30))
industry_fund(data_temp31.tail(30)["holdingTicker"],weight,myreportDate,u"当季综合股混后30基金行业配置:")
# for i in data_temp31.iloc[-11:-1]["holdingTicker"]:
#     analysis_stock(i)
print("总数量"),
print(len(data_temp31))
print(data_temp31.to_html())
```
