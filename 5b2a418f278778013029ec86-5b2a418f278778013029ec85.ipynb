```python
import numpy as np
import pandas as pd
from pandas import Series
import matplotlib.pyplot as plt
pd.options.display.width = 2000
a=[0.05,0.10]
mean=Series(a)
cov=[[0.0025,-0.01],
 [-0.01,0.04]]
port_returns = []
port_variance = []
for p in range(4000):
        weights = np.random.random(2)
        weights /=np.sum(weights)
        port_returns.append(np.sum(mean*weights))
        port_variance.append(np.sqrt(np.dot(weights.T, np.dot(cov, weights))))
port_returns1 = np.array(port_returns)
port_variance1= np.array(port_variance)
plt.figure(figsize = (32,16))
plt.scatter(port_variance1, port_returns1,color='b', marker = 'o')

#
a=[0.05, 0.10]
mean=Series(a)
cov=[[0.0025,0.00],[0.00,0.04]]
port_returns = []
port_variance = []
for p in range(4000):
    weights = np.random.random(2)
    weights /=np.sum(weights)
    port_returns.append(np.sum(mean*weights))
    port_variance.append(np.sqrt(np.dot(weights.T, np.dot(cov, weights))))
port_returns2 = np.array(port_returns)
port_variance2 = np.array(port_variance)
plt.figure(figsize = (32,16))
plt.scatter(port_variance2, port_returns2,color='r', marker = 'o')

#
a=[0.05,0.10]
mean=Series(a)
cov=[[0.0025, 0.01],
 [0.01, 0.04]]
port_returns = []
port_variance = []
for p in range(4000):
        weights = np.random.random(2)
        weights /=np.sum(weights)
        port_returns.append(np.sum(mean*weights))
        port_variance.append(np.sqrt(np.dot(weights.T, np.dot(cov, weights))))
port_returns3 = np.array(port_returns)
port_variance3 = np.array(port_variance)
plt.figure(figsize = (32,16))
print plt.scatter(port_variance3, port_returns3,color='b',marker='o')
#
print plt.plot(port_variance1, port_returns1,'ks',port_variance2, port_returns2,'ks',port_variance3, port_returns3,color='b',marker='o')  
```
