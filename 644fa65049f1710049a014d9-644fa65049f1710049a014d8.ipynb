在柯蒂斯 费思的著作《海龟交易法则》中，把交易市场划分为下面四个状态：
	稳定平静：价格在一个相对较小的范围内上下波动，很少超出这个范围。
	稳定波动：有较大的日间变化，但是没有重大的月际变化。
	平静的趋势：价格在几个月中呈现出缓慢的运动或者趋势，但是并没有明显的回调和反方向运动。
	波动的趋势：价格有较大的变化，偶尔伴随着剧烈的短期反向波动。
    
    为何区分不是很明显？是不是程序有bug?
    但看中国平安 
    2023-04-27  平静的趋势  201  
    2023-04-28  平静的趋势  202  

```python
def cal_er(ser):
    '''输入数据ser为价格的Series'''
    #print ser-ser.shift(1)
    a=abs(ser.iloc[-1]-ser.iloc[0])
    b=abs(ser-ser.shift(1)).sum()
    return a/b
```

```python
import pandas as pd
import talib as tb
start_date = "2007-01-01"  # 开始日期
end_date = "2020-12-31"  # 结束日期
start_date = "2022-07-01"  # 开始日期
end_date = "2023-04-29"  # 结束日期
df=DataAPI.MktEqudAdjGet(ticker=u"601318",beginDate=start_date,endDate=end_date,isOpen="1",field=u"",pandas="1")
print df.columns
#df.rename(columns={'closePrice':'close'}, inplace = True)
df.index=df.tradeDate
N=30
ATR = tb.ATR(df['highestPrice'].values, df['lowestPrice'].values, df['closePrice'].values, timeperiod=N)

df["atr"]=ATR
df=df[["ticker","secShortName","closePrice","atr"]]
df["close14"]=df["closePrice"].shift(N)
df["mt"]=df.apply(lambda x :(x.closePrice-x.close14)/x.atr, axis=1)
df.loc[df.mt>2,"trending"]=1
df["er"]=df["closePrice"].rolling(window=14).apply(lambda x:cal_er(pd.Series(x)))
#con1=(df.mt<2)&(df.er<0.5)
#con2=(df.mt<2)&(df.er>=0.5)
#con3=(df.mt>=2)&(df.er<0.5)
#con4=(df.mt>=2)&(df.er>=0.5)
thr=1.5
con1=(abs(df.mt)<thr)&(df.er<0.5)
con2=(abs(df.mt)<thr)&(df.er>=0.5)
#下降趋势
#con11=(df.mt<=-2)&(df.mt<2)&(df.er<0.5)
#con21=(df.mt<=-2)&(df.mt<2)&(df.er>=0.5)
con3=(abs(df.mt)>=thr)&(df.er<0.5)
con4=(abs(df.mt)>=thr)&(df.er>=0.5)
df.loc[con1,"市场状态"]="稳定平静"
df.loc[con2,"市场状态"]="稳定波动"
#df.loc[con11,"市场状态"]="平静的下趋势"
#df.loc[con21,"市场状态"]="波动的下趋势"
df.loc[con3,"市场状态"]="平静的趋势"
df.loc[con4,"市场状态"]="波动的趋势"
#print df[df.mt>2]
df["day"]=range(len(df))
df=df[['secShortName',"closePrice","close14","atr","mt","er","市场状态","day"]]
print df
```
