## 3. 二八轮动策略的时间窗口选择

----

我们看到的二八轮动策略中，都是采用最近二十日收益作为选择标准，可能基于经验判断这是最适合A股的策略；下面我们基于回测验证这一点：

```python
import pandas as pd
import numpy as np
from CAL.PyCAL import *

# 可编辑部分与 strategy 模式一样，其余部分按本例代码编写即可

# -----------回测参数部分开始，可编辑------------
start = '2013-05-01'
end = u'2016-09-12'
benchmark = 'HS300'
universe = ['510300.XSHG',   # 300 ETF 代替大盘蓝筹板块
            '159915.XSHE',   # 创业板 ETF 代替中小盘板块
            '511880.XSHG']   # 银华日利
capital_base = 100000

# ---------------回测参数部分结束----------------

# 把回测参数封装到 SimulationParameters 中，供 quick_backtest 使用
sim_params = quartz.SimulationParameters(start, end, benchmark, universe, capital_base)
# 获取回测行情数据
idxmap, data = quartz.get_daily_data(sim_params)
# 运行结果
results = {}
# 调整参数快速回测
for window in range(4, 30, 1):
    
    # ---------------策略逻辑部分----------------
    refresh_rate = 1
    commission = Commission(0.0002, 0.0002)
    
    def initialize(account):
        account.fund_2 = universe[0]   # 二， 表示蓝筹大盘股
        account.fund_8 = universe[1]   # 八， 表示中小盘股票
        account.fund_0 = universe[2]   # 黄金，债基，货基等熊市保守投资

    def handle_data(account):
        close_prices = account.get_attribute_history('closePrice', window + 1)  # 拿取过去window+1日收盘价，window一般为二十日
        price_2 = close_prices[account.fund_2]
        price_8 = close_prices[account.fund_8]

        signal_2 = price_2[-1]/price_2[0] - 1.0    # 判断大盘蓝筹基金二十日涨跌
        signal_8 = price_8[-1]/price_8[0] - 1.0    # 判断中小盘基金二十日涨跌

        if signal_2 < 0 and signal_8 < 0:    # 大盘蓝筹基金和中小盘基金二十日涨跌幅为负
            order_pct_to(account.fund_2, 0)
            order_pct_to(account.fund_8, 0)
            order_pct_to(account.fund_0, 0.99)
        else:
            if signal_2 > signal_8:          # 大盘蓝筹二十日涨，买入蓝筹基金
                order_pct_to(account.fund_8, 0)
                order_pct_to(account.fund_0, 0)
                order_pct_to(account.fund_2, 0.99)
            if signal_8 > signal_2:          # 中小盘二十日涨，买入中小盘基金
                order_pct_to(account.fund_2, 0)
                order_pct_to(account.fund_0, 0)
                order_pct_to(account.fund_8, 0.99) 
    # ---------------策略逻辑部分结束----------------

    # 把回测逻辑封装到 TradingStrategy 中，供 quick_backtest 使用
    strategy = quartz.TradingStrategy(initialize, handle_data)
    # 回测部分
    bt, acct = quartz.quick_backtest(sim_params, strategy, idxmap, data, refresh_rate=refresh_rate, commission=commission)

    # 对于回测的结果，可以通过 perf_parse 函数计算风险指标
    perf = quartz.perf_parse(bt, acct)

    # 保存运行结果
    tmp = {}
    tmp['annualized_return'] = perf['annualized_return']
    tmp['volatility'] = perf['volatility']
    tmp['max_drawdown'] = perf['max_drawdown']
    tmp['alpha'] = perf['alpha']
    tmp['beta'] = perf['beta']
    tmp['sharpe'] = perf['sharpe']
    tmp['information_ratio'] = perf['information_ratio']
    
    results[window] = tmp
    print str(window),
print 'done'
```

```python
# results 转换为 DataFrame
import pandas
results_pd = pandas.DataFrame(results).T.sort_index()
results_pd
```

##### 从上表可以看出，窗口设置为20天，除了得到更多的收益之外，其信息比率、夏普比率、最大回撤等风险指标也是综合表现最好的。
