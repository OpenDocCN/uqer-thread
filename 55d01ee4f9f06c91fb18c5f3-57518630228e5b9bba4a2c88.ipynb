```python
#coding=utf-8
"""
    uqerAPI.py
    对DataAPI进行包装，方便在notebook中研究使用。
"""

import pandas as pd
import datetime
# DataAPI和quartz在notebook是预先加载，library里需要显示的import。
import DataAPI
from quartz.api import set_universe

def get_industry_stock(universe='HS300', industry=[]):
    """获取指定行业包含的股票列表。
    :param universe: 股票池
    :param industry: 行业列表
    
    Returns: 行业为key，股票列表为value的字典"""
    
    df = pd.DataFrame(columns=['secID', 'industryName1'])
    tickers = [e[:6] for e in set_universe(universe)]

    for i in xrange(0, len(tickers), 5):
        # 在帮助-DataAPI使用帮助中找到一个行业相关的API：EquIndustryGet，
        # EquIndustryGet每次只能获取五只股票信息，待改进。
        ticker_str = ','.join(tickers[i:min(len(tickers), i+5)])
        
        tmp_df = DataAPI.EquIndustryGet(industry=u'',
                                        secID=u'',
                                        ticker=ticker_str,
                                        industryVersionCD=u'010303',
                                        intoDate=datetime.datetime.now().strftime('%Y%m%d'),
                                        field=u'secID,industryName1',
                                        pandas='1')
        
        df = df.append(tmp_df,ignore_index=True)
    
    # 默认industry为空，返回所有行业数据，如果有指定industry，则返回指定的industry。
    gp = df.groupby('industryName1')['secID']
    return {e[0]: list(e[1].values) for e in list(gp) if e[0] in industry or not industry}

# get_industry_stock(industry=[])
```
