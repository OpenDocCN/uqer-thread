# 再看宏观因子择时——Online learning

# `思路整理`
***

- #### 筛选对于市场波动有一定预测能力的指标，备选：
    - #### 实体经济指标：`PMI(采购经理人指数)`
    - #### 物价水平指标：`CPI(消费者价格指数)`、`PPI(生产者价格指数)`
    - #### 货币财政指标：`M1(狭义货币量)`、`M2(广义货币量)`
    - #### 居民消费能力：`Inc1(城镇居民实际可支配收入同比)`、`Inc2(农村居民实际可支配收入同比)`


- #### Online Learning，每期采用当期能够获得的所有信息拟合模型，预测下一期市场收益

- #### 利用所得预测收益进行操作，若$E(r_{t+1})$大于一定门槛值则持有沪深300ETF，否则空仓。



# `所用模型`
***

\[R_t = \beta_0 +\beta_1 PMI_{t-1}+\beta_2 CPI_{t-1}+ \beta_3PPI_{t-1}+\beta_4 M1_{t-1}+\beta_5 M2_{t-1}+\beta_6 Inc1_{t-1}+\beta_7 Inc2_{t-1} + \epsilon_t\]
- #### 参数从左至右分别为：$t$月股指收益率、$t-1$月PMI同比增长率、$t-1$月CPI同比增长率、$t-1$月PPI同比增长率、$t-1$月M1同比增长率、$t-1$月M2同比增长率、$t-1$月城镇居民实际可支配收入同比、$t-1$月农村居民实际可支配收入同比；

- #### 滞后期选择是根据数据的获得时间，PMI数据在每月月初公布，而其他宏观数据在每月月中公布。另外，城村居民实际可支配收入每季公布，在这里调整至月度，每月采用可获得的最近公布值。

- #### 引入工具包并定义全局设定：

```python
# libs
from __future__ import division
import seaborn as sns
import datetime as dt
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
import string
from sklearn import linear_model
from CAL.PyCAL import *

# settings
sns.set_style('whitegrid')
__fsize = (12, 12*0.618)
__today = dt.datetime.today().strftime('%Y%m%d')
DataAPI.settings.cache_enabled = True
```

```python
def monthBefore(year = 2016, month = 1, delta = 2):
    # 获取delta个月之前的年月
    rtnMonth = (month + 12 - (delta%12)) % 12
    rtnMonth = 12 if rtnMonth == 0 else rtnMonth
    yearDelta = int((delta - month) / 12) + 1
    rtnYear = year - yearDelta if delta >= month else year
    return dt.datetime(rtnYear, rtnMonth, 1).strftime('%Y-%m')
```

- #### 用到的宏观数据指标是：

```python
DataAPI.ChinaMacroInfoGet(indicID = ['M030000157', 'M030000012', 'M100000004', 'M100000005', 'M020000008', 'M040000013', 'M100000007', 'M100000008', 'M060000005', 'M060000021'])
```

- #### 获取每月指数行情数据：

```python
cal = DataAPI.TradeCalGet(exchangeCD=u"XSHE",beginDate= '20050701',endDate = __today , field= 'calendarDate,isMonthEnd')
cal = cal[cal.isMonthEnd == 1]
cal['stockIndex'] = map(lambda x: DataAPI.MktIdxdGet(tradeDate= x, ticker = '000001', 
                                                     field = 'closeIndex').values[0][0], cal.calendarDate.values)
cal.calendarDate = cal.calendarDate.apply(lambda x: x[:7])
cal = cal.set_index('calendarDate').drop('isMonthEnd', axis = 1)
```

- #### 数据组合与处理：

```python
macroData = DataAPI.ChinaMacroDataGet(indicID = 
            ['M030000157', 'M030000012', 'M100000004', 'M100000005', 'M020000008', 'M040000013', 'M100000007', 'M100000008', 'M060000005', 'M060000021'], 
            beginDate= '20060101').pivot(index = 'periodDate', values = 'dataValue', columns = 'indicName')
macroData.columns = ['Inc1_SOY', 'Inc2_SOY', 'PMI', 'CPI_MOM', 'PPI_MOM', 'VAI_SS_MOM', 'M1', 'M2', 'M1_MOY', 'M2_MOY']
def lastUpdate(n,series):
    if n == 0:
        return series[n]
    if not np.isnan(series[n]):
        return series[n]
    else:
        return lastUpdate(n-1, series)
    
macroData['CPI'] = 1 + macroData.CPI_MOM / 100
macroData['CPI'] = macroData.CPI.cumprod()
macroData['PPI'] = 1 + macroData.PPI_MOM / 100
macroData['PPI'] = macroData.PPI.cumprod()
macroData['M1_index'] = macroData.M1 / macroData.M1[:1][0]
macroData['M2_index'] = macroData.M2 / macroData.M2[:1][0]
macroData['M2mM1_index'] = macroData.M2 - macroData.M1
macroData['M2mM1_index'] = macroData['M2mM1_index'] / macroData['M2mM1_index'][:1][0]

macroData['CPI_MOY'] = macroData.CPI
macroData['PPI_MOY'] = macroData.PPI
macroData['PMI_MOY'] = macroData.PMI

# macroData['CPI_MOM'] = macroData.CPI
# macroData['PPI_MOM'] = macroData.PPI
# macroData['PMI_MOM'] = macroData.PMI
# macroData['M1_MOM'] = macroData.M1_index
# macroData['M2_MOM'] = macroData.M2_index

for i in range(len(macroData)):
    macroData.Inc1_SOY[i] = lastUpdate(i, macroData.Inc1_SOY) if np.isnan(macroData.Inc1_SOY[i]) else macroData.Inc1_SOY[i]
    macroData.Inc2_SOY[i] = lastUpdate(i, macroData.Inc2_SOY) if np.isnan(macroData.Inc2_SOY[i]) else macroData.Inc2_SOY[i]
    macroData.CPI_MOY[i] = (macroData.CPI_MOY[i] / macroData.CPI[i-12] - 1)*100 if (i-12>=0) else np.nan
    macroData.PPI_MOY[i] = (macroData.PPI_MOY[i] / macroData.PPI[i-12] - 1)*100 if (i-12>=0) else np.nan
    macroData.PMI_MOY[i] = (macroData.PMI_MOY[i] / macroData.PMI[i-12] - 1)*100 if (i-12>=0) else np.nan

mData = macroData[['Inc1_SOY', 'Inc2_SOY', 'M1_MOY', 'M2_MOY', 'CPI_MOY', 'PPI_MOY', 'PMI_MOY']]
mData.index = map(lambda x: x[:7], mData.index)
tmData = mData.merge(cal, left_index = True, right_index = True)
tmData.stockIndex = tmData.stockIndex / tmData.stockIndex[:1][0]
tmData['indexReturn'] = tmData.stockIndex.pct_change()*100

tmData.tail(5)
```

```python
regData = tmData[['Inc1_SOY', 'Inc2_SOY', 'M1_MOY', 'M2_MOY', 'CPI_MOY', 'PPI_MOY', 'PMI_MOY', 'indexReturn']]
regData.indexReturn[:-1] = tmData.indexReturn[1:]; regData = regData[:-1]

regData.dropna(inplace = True)
regr = linear_model.LinearRegression()
regr.fit(regData[['Inc1_SOY', 'Inc2_SOY', 'CPI_MOY', 'PPI_MOY', 'M1_MOY', 'M2_MOY', 'PMI_MOY']], regData.indexReturn)
# print "R-square of this regression is %f."%regr.score(regData[['Inc1_SOY', 'Inc2_SOY', 'CPI_MOY', 'PPI_MOY', 'M1_MOY', 'M2_MOY', 'PMI_MOY']], regData.indexReturn)
regData['predictReturn'] = regr.predict(regData[['Inc1_SOY', 'Inc2_SOY', 'CPI_MOY', 'PPI_MOY', 'M1_MOY', 'M2_MOY', 'PMI_MOY']])
sns.pairplot(regData[['indexReturn', 'predictReturn']], x_vars= 'predictReturn', y_vars= 'indexReturn', kind = 'reg', size= 7)
print "符号准确率是%f."%(len(regData[regData.predictReturn*regData.indexReturn > 0]) / len(regData))
```
