**Alpha#54**
$$ \large((-1 * ((low - close) * (open^5))) / ((low - high) * (close^5)))$$



观察公式发现，仅用到了价格数据。而且比较奇怪的是，我们知道，如果k线不是一字板且有下影线的时候，上式始终小于0！ 

如果没有下影线，说明close = low或者open = low，第一种情况值为0，第二种情况值还是小于0的！

如果是一字板，式子算出来应该是nan，被舍弃了。

既然值恒小于等于0 ，我们希望买入的是分母小、分子大的股票。 最优的情况是昨天上涨，没有下影线，涨幅较大，没有上影线或最高价较高。

我觉得这是利用**动量**的思想，也是**追涨**的思想。 

为什么要5次方？ 为了拉开差距，只是一种处理手段，正是因为这个，我们才偏向昨日上涨的股票~

但是权重为负数，我怀疑这是不是作者的本意？？难道是多写了个取反？？
