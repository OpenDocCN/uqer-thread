**现象**：
在测试周期里面，每年6，12月份中期会有几十只股票从沪深300里面去除，也有些股票进入。算下来这段时间内将近有半数沪深300股票出现变动。如果按照生存者偏差来模拟的话，在模拟过程中等权投入HS300，更新时候也随即更新的话，基本上收益曲线和沪深300经过取权重的曲线差不多，然后如果是用现在的沪深300的股票模拟的话，从回测2012年到现在4年半时间，设置False时收益也就是现在用现在沪深300股票时期末收益是135%，设置True时也就是模拟测试当天沪深300成份股时期末收益是36%，差异是非常巨大的，如果有人用了set_universe('HS300')的话建议重新进行测试，结果会差很多。

**结论**：
1. 如果有人在策略里面用了set_universe('HS300')，而且我代码没有写错的话，建议重新修改代码重新测试，因为回测肯定失真，无法对策略产生正确预估，风险太大。
2. 参考了一个沪深300的等权策略，实际上所有收益来自于生存者偏差，不足信，实际上等权买入只比HS300强非常有限的一小部分

**用法**：test300设置为True或者False，运行2次对比结果。
1. test300=True时为模拟回测当天的沪深300成份股情况，如果是True收益曲线基本就是HS300，如果是False就是按照现在
