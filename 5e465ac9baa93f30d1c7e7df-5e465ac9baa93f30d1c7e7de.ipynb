# 最大回撤

```python
import datetime

# 最大回撤率
# ticker:股票交易代码，date_start:开始计算日期‘20180213’，date_end:截至日期‘20180217’，time_span:窗口天数
def stock_huichelv(ticker,date_start,date_end,time_span=0):
    ticker = str(ticker)
    date_start = str(date_start)
    date_end = str(date_end)

    # 调用接口获取数据
    data = DataAPI.MktEqudGet(secID=u"",ticker=ticker,tradeDate=u"",beginDate=date_start,endDate=date_end,isOpen="",
                   field=u"",pandas="1")
    data_count = data['secID'].count()-1 # 数据总数作为循环的范围（和日期长度不一样，因为休息日没有数据）
    # 取出数据集中的收盘价和交易日期两列
    deal_data = data.ix[:,['closePrice','tradeDate']]
    day_max =-100
    day_date_start = 0
    day_date_end = 0
    # 以窗口跨度每次向后移动一天
    for t in range(data_count-time_span):
        # 循环窗口内每天对比之后的所有数据
        # 1与30天内，2与29天内，。。。。
        if time_span == 0:
            for j in range(t,data_count):
                    price_out = deal_data.iloc[t][0]
                    price_in = deal_data.iloc[j][0]
                    if price_in > price_out:
                        continue
                    priceChgPct = (price_out - price_in)/price_out*100
                    if priceChgPct > day_max:
                        day_max = priceChgPct
                        day_date_start = deal_data.iloc[t][1]
                        day_date_end = deal_data.iloc[j][1] 
        else:
            for i in range(t,time_span+t):
                # 1与1、1与2、1与3.。。。中最大撤回率
                for j in range(i,time_span+t):
                    price_out = deal_data.iloc[i][0]
                    price_in = deal_data.iloc[j][0]
                    if price_in > price_out:
                        continue
                    priceChgPct = (price_out - price_in)/price_out*100
                    if priceChgPct > day_max:
                        day_max = priceChgPct
                        day_date_start = deal_data.iloc[i][1]
                        day_date_end = deal_data.iloc[j][1]
    print(day_max)
    print(day_date_start,day_date_end)
```

```python
# 1、以月为窗口，挑选一只基金or股票，查看它近2年最大回撤率
# 贵州茅台近两年最大回车率
stock_huichelv(600519,20180213,20200213,30) 
```
