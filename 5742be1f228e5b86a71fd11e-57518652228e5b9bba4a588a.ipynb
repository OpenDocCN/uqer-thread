# 算法交易入门（02：双均线策略）

双均线策略:以5日均线穿越20日均线买入，反之，卖出。以中国平安（000001.XSHE）为例，每次买入资本金固定。本练习重点要掌握构造策略的过程。

```python
import numpy as np
import pandas as pd
from pandas import DataFrame

import datetime
import matplotlib.pyplot as plt
```

```python
#回测区间
start = '20120801'                       # 回测起始时间
end   = '20150801'                         # 回测结束时间

universe = [ '000001.XSHE']

#获取回测期间全部数据
bars = DataFrame()
bars = DataAPI.MktEqudAdjGet(secID=universe[0],beginDate=start,endDate=end,field=u"tradeDate,openPrice,highestPrice,lowestPrice,closePrice,turnoverVol",pandas="1")
bars = bars.rename(columns={'tradeDate':'Date','openPrice':'Open','highestPrice':'High','lowestPrice':'Low','closePrice':'Close','turnoverVol':'Volume'})
bars['Date'] =  pd.to_datetime(bars['Date']) #因为DataAPI所有日期都是以字符串形式存储的，因此此处需要转换成日期时间型,这里很重要！
bars = bars.set_index('Date')#转换后index是DatetimeIndex型
bars.head()
```

```python
#产生信号
def generate_signals(bars,short_window=5,long_window=20):

    """Returns the DataFrame of symbols containing the signals
    to go long, short or hold (1, -1 or 0)."""
    signals = DataFrame()
    signals['Close'] = bars.Close
    signals['signal'] = 0.0

    # Create the set of short and long simple moving averages over the 
    # respective periods
    signals['short_mavg'] = pd.rolling_mean(signals['Close'], short_window, min_periods=1)
    signals['long_mavg'] = pd.rolling_mean(signals['Close'], long_window,min_periods=1)

    # Create a 'signal' (invested or not invested) when the short moving average crosses the long
    # moving average, but only for the period greater than the shortest moving average window
    signals['signal'][short_window:] = np.where(signals['short_mavg'][short_window:] 
        > signals['long_mavg'][short_window:], 1.0, 0.0)   

    # Take the difference of the signals in order to generate actual trading orders
    signals['positions'] = signals['signal'].diff()   
    return signals

short_window=5
long_window=20

signals = generate_signals(bars,short_window,long_window)
signals[:25]
```

```python
#定义头寸
def generate_positions(initial_capital,signals):
    positions = pd.DataFrame(index=signals.index).fillna(0.0)
    symbol = universe[0]
    signals[symbol] =  signals.apply(lambda row: int(initial_capital/row['Close']/100)*100, axis = 1)  # 以固定资本金买入
    positions[symbol] = signals[symbol]*signals['signal']  
    return positions
```

```python
#计算组合
def portfolio(initial_capital,positions,bars):
    portfolio = positions*bars['Close']
    pos_diff = positions.diff()

    portfolio['holdings'] = (positions*bars['Close']).sum(axis=1)
    portfolio['cash'] = initial_capital - (pos_diff*bars['Close']).sum(axis=1).cumsum()

    portfolio['total'] = portfolio['cash'] + portfolio['holdings']
    portfolio['returns'] = portfolio['total'].pct_change()
    return portfolio

initial_capital = 100000
positions = generate_positions(initial_capital,signals)
returns = portfolio(initial_capital,positions,bars)
returns.tail(10)
```

```python
#研究策略性能

#夏普率公式
def create_sharpe_ratio(returns, periods=250):
    """
    Create the Sharpe ratio for the strategy, based on a 
    benchmark of zero (i.e. no risk-free rate information).

    Parameters:
    returns - A pandas Series representing period percentage returns.
    periods - Daily (250), Hourly (250*4), Minutely(250*4*60) etc.
    """
    return np.sqrt(periods) * (np.mean(returns)) / np.std(returns)

#最大回撤公式
def create_drawdowns(equity_curve):
    """
    Calculate the largest peak-to-trough drawdown of the PnL curve
    as well as the duration of the drawdown. Requires that the 
    pnl_returns is a pandas Series.

    Parameters:
    pnl - A pandas Series representing period percentage returns.

    Returns:
    drawdown, duration - Highest peak-to-trough drawdown and duration.
    """

    # Calculate the cumulative returns curve 
    # and set up the High Water Mark
    # Then create the drawdown and duration series
    hwm = [0]
    eq_idx = equity_curve.index
    drawdown = pd.Series(index = eq_idx)
    duration = pd.Series(index = eq_idx)

    # Loop over the index range
    for t in range(1, len(eq_idx)):
        cur_hwm = max(hwm[t-1], equity_curve[t])
        hwm.append(cur_hwm)
        drawdown[t]= hwm[t] - equity_curve[t]
        duration[t]= 0 if drawdown[t] == 0 else duration[t-1] + 1
    return drawdown.max(), int(duration.max())
```

```python
strsr = create_sharpe_ratio(returns.returns,250)
print "Strategy Sharpe ratio: %s" %strsr
maxDrawdown,maxDrawdownDuration = create_drawdowns(returns.returns) # 最大回撤，最长回撤期
print "maxDrawdown:%s" %maxDrawdown
print "maxDrawdownDuration: %s" %maxDrawdownDuration
```
