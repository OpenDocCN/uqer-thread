# 通联因子有效性测试-面板分析

> 在这个简单的示例中，将尝试使用最简单的线性回归模型去验证通联数据提供的因子数据的有效性。这次的分析与姐妹篇[通联因子有效性测试](https://uqer.datayes.com/community/share/56c3fce0228e5b0fe3b17bf2)的区别主要是本篇不是**时间序列分析**，而是基于**面板分析**。

# *测试原理*

* 使用因子最近3个月的变化值预测下面3个月的收益率；

* 因子的变化值会在``universe``上做标准化，收益率也会做相似的操作；

* 测试范围为随机选择的50只股票（可以扩展至全市场）；

* 选取的因子是通联数据api提供的所以因子。

## 1. 准备

首先我们引入需要使用的包

```python
import datetime as dt
import pandas as pd
import numpy as np
from scipy import stats

from quartz.api import set_universe
from CAL.PyCAL import *
```

## 2. 计算日期序列

回测周期我们打算从2008年开始，并且以3个月（季度）为周期。这里我们使用``pandas``生成季度末的日期序列。除此以外，我们需要把这些日期调整到交易所最近的交易日，这里我们又使用了``CAL``。

```python
cal = Calendar('China.SSE')

time_series = pd.date_range('2008-1-1', '2016-2-17', freq='3M')
biz_date_list = []
for date in time_series:
    date = cal.adjustDate(date, BizDayConvention.Preceding)
    biz_date_list.append(date.toDateTime())
```

## 3. 确定回测``universe``以及因子范围

* 我们选取了全A股中按照名称排序前50位的股票（当然也可以直接回测全A股，这样耗时会很长很长~）；

* 因子方面，我们选取了现在``MktStockFactorsOneDayGet``中全部因子，现在的数量为133个。

```python
universe = set_universe('A')[:50]
start = '20080101'
end = '20160217'

def filter_func(x):
    if str.isupper(x[0]):
        return True
    else:
        return False

factors = filter(filter_func, dir(Factor))
```

## 4. 回测

回测过程是简单的线性回归的过程。用每季度因子的变化值与下个季度的股票价格变化进行回归。我们最后收集的统计数据包括：

* 回归系数；

* 回归结果的p值。

最后对于每个因子，我们计算所有股票回归结果的：

* 回归系数的均值以及中值；

* p值的均值以及中值。

```python
slopes_mean = []
slopes_median = []
pvalues_mean = []
pvalues_median = []
secNum = []

for i, factor in enumerate(factors):

    slopes = []
    pvalues = []
    
    previous_previous_factor_data = pd.DataFrame()
    previous_factor_data = pd.DataFrame()
    previous_price_data = pd.DataFrame()

    for date in biz_date_list:
        # 获取特定股票该因子在特定交易日的值的值
        factor_data =DataAPI.MktStockFactorsOneDayGet(date.strftime('%Y%m%d'), secID=universe, field=['secID',factor])
        factor_data.dropna(inplace=True)
        factor_data.set_index('secID', inplace=True)

        # 获取所有股票价格在特定交易日的值
        price_data = DataAPI.MktEqudAdjGet(date.strftime('%Y%m%d'), secID=universe,  field=['secID', 'closePrice'])
        price_data.set_index('secID', inplace=True)
        
        if previous_previous_factor_data.empty or previous_factor_data.empty:
            previous_previous_factor_data = previous_factor_data
            previous_factor_data = factor_data
            previous_price_data = price_data
            continue
            
        # 计算差值
        previous_factor_change = np.log(previous_factor_data / previous_previous_factor_data)
        price_change = np.log(price_data / previous_price_data)

        # 数据合并
        merged_data = pd.merge(previous_factor_change, price_change, left_index=True, right_index=True)
        merged_data.dropna(inplace=True)

        # 进行标准化
        diff_series = merged_data.diff().dropna()
        merged_data[factor] = (merged_data[factor] - merged_data[factor].mean()) / merged_data[factor].std()
        merged_data['closePrice'] = (merged_data['closePrice']   - merged_data['closePrice'].mean())/ merged_data['closePrice'].std()
        
        # 进行回归的序列。注意，因子变化是当季度值，价格变化是下一季度的值。
        x_train = merged_data[factor].values
        y_train = merged_data['closePrice'].values

        # 进行回归， 有效数据必须超过10个股票
        if len(x_train) >= 10:
            regr = stats.linregress(x_train, y_train)
            slope, intercept, r_value, p_value, std_err = stats.linregress(x_train, y_train)
            if not np.isnan(slope):
                slopes.append(slope)
                pvalues.append(p_value)
                
        previous_previous_factor_data = previous_factor_data
        previous_factor_data = factor_data
        previous_price_data = price_data
    
    slopes_mean.append(np.mean(slopes))
    slopes_median.append(np.median(slopes))
    pvalues_mean.append(np.mean(pvalues))
    pvalues_median.append(np.median(pvalues))
    secNum.append(len(slopes))
    
    print(u"{0}. 因子 {1} 测试完成！".format(i, factor))

    # 生成报告
    report = pd.DataFrame({u'因子': factors[:i+1],
                           u'系数（均值）': slopes_mean,
                           u'系数（中值）': slopes_median,
                           u'p值（均值）': pvalues_mean,
                           u'p值（中值）': pvalues_median,
                           u'有效日数': secNum})

report = report[[u'因子', u'有效日数', u'系数（均值）', u'系数（中值）', u'p值（均值）', u'p值（中值）']]
```

# 5. 结果

我们最后按照``p(中值)``的大小排序回顾结果。理论上讲，p值越小的情况下，回归结果越显著。
