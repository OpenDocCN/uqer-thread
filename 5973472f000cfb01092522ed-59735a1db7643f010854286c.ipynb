### 一、定义变量

```python
import numpy as np

"""
total_money 现金总额
buy_stock_num 买入股数
start_price 买入价
end_price 卖出价
price_chng 每日相对买入是收益增幅
index_chng 基准相对买入时收益增幅
cost_rate 手续费率
hold_days 持有天数
safe_rate 无风险利率
"""
total_money = 20000
buy_stock_num = 100
start_price = 163.33
end_price = 163.93
daily_total_return = np.array([0, 0.0547, 0.0345, 0.0135, 0.0084, 0.0022])
bench_daily_total_return = np.array([0, 0.0305, 0.0304, 0.0312, 0.0072, 0.0037])
cost_rate = 0.001
hold_days = 5
safe_rate = 0.04
```

### 二、收益

#### 公式
- 基本收益公式：
$$ P = \dfrac{P_{end} -P_{start}}{P_{end}} * 100\% $$ 
$$ P_{start}  = 初始账户总金额 = 投入金额+账户现金$$
$$ P_{end} = 卖出金额+账户现金 $$
- 年化收益公式: 
$$ R_p = ((1 + P)^{250/n} -1 ) * 100\% $$
$$ P = 策略收益 $$
$$ n = 策略运行天数 $$
- 比较基准年化
$$ R_m = ((1+P)^{250/n}-1) * 100\% $$
$$ P = 基准开始指数/基准结束指数 $$
$$ n = 基准运行天数 $$

#### 计算
计算策略基本收益率、年化收益率与基准年化收益。

```python
# 买入金额
buy_money = buy_stock_num * start_price
# 费用
cost_money = buy_money * (1+cost_rate)
# 剩余现金
# remain_money = total_money - buy_money
remain_money = total_money - cost_money
```

```python
# 卖出金额
sale_money = buy_stock_num * end_price
# 当前所有金额
now_total_money = sale_money + remain_money
```

```python
# 策略基本收益
total_return = (now_total_money - total_money)/total_money
# 策略年化收益
annualized_return = np.power((now_total_money/total_money), 250.0/hold_days) - 1
# 比较基准年化
bench_return = np.power((1+bench_chng[-1]), 250.0/hold_days) - 1
```

```python
print "策略基本收益为", total_return
print "策略年化收益为", annualized_return
print "基准年化收益为", bench_return
```

### 三、贝塔 Beta

#### 描述
表示投资的系统性风险，反映了策略对大盘变化的敏感性。

例如一个策略的Beta为1.3，则大盘涨1%的时候，策略可能涨1.3%，反之亦然；如果一个策略的Beta为-1.3，说明大盘涨1%的时候，策略可能跌1.3%，反之亦然。

#### 公式

$$ Beta = \beta_p = \dfrac{Cov(D_p, D_m)}{Var(D_m)} $$
$$ D_p = 策略每日收益 $$
$$ D_m = 基准每日收益 $$
$$ Cov(D_p, D_m) = 策略每日收益与基准每日收益的协方差 $$
$$ Var(D_m) = 基准每日收益的方差 $$

#### 解释
β<0	投资组合和基准的走向通常反方向，如空头头寸类
β=0	投资组合和基准的走向没有相关性，如固定收益类
0<β<1  投资组合和基准的走向相同，但是比基准的移动幅度更小
β=1	投资组合和基准的走向相同，并且和基准的移动幅度贴近
β>1	投资组合和基准的走向相同，但是比基准的移动幅度更大

#### 计算策略Beta值

** 计算每日收益 **

```python
# 1. 计算策略每日收益
daily_return = np.diff(daily_total_return)/(1+daily_total_return[:-1])
# 2. 计算基准每日收益
bench_daily_return = np.diff(bench_daily_total_return)/(1+bench_daily_total_return[:-1])
# 3. 计算策略每日收益与基准每日收益协方差
cov = ((daily_return - daily_return.mean()) * (bench_daily_return - bench_daily_return.mean())).sum()/hold_days
# 4. 计算基准每日收益方差
bench_var = bench_daily_return.var()
# 5. 计算beta值
beta = cov/bench_var
print "策略beta值为", beta
```

### 四、阿尔法 Alpha

#### 描述
投资中面临着系统性风险（即Beta）和非系统性风险（即Alpha），Alpha是投资者获得与市场波动无关的回报。比如投资者获得了15%的回报，其基准获得了10%的回报，那么Alpha或者价值增值的部分就是5%。

#### 公式
$$ Alpha =\alpha = R_p - [R_f + \beta_p(R_m - R_f)] $$
$$ R_p = 策略年化收益 $$
$$ R_m = 基准年化收益 $$
$$ R_f = 无风险利率 (默认0.04) $$
$$ \beta_p = 策略beta值 $$

#### 解释
α>0	策略相对于风险，获得了超额收益
α=0	策略相对于风险，获得了适当收益
α<0	策略相对于风险，获得了较少收益

#### 计算策略Alpha值

```python
alpha = annualized_return - (safe_rate + beta * (bench_return - safe_rate))
print "策略alpha值为", alpha
```

### 五、波动率 Volatility

#### 描述
用来测量风险性，波动越大代表策略风险越高

#### 公式
$$ Volatility = \sigma_p = \sqrt{\dfrac{250}{n-1}\sum_i^n(r_p-\bar{r}_p)^2}$$
$$ r_p = 每日收益率 $$
$$ \bar{r}_p = 每日收益率平均值 \dfrac{1}{n}\sum_{i=1}^{n}r_p$$

#### 计算策略与基准的波动率

```python
# 1. 计算策略的波动率
volatility = daily_return.std(ddof=1) * np.power(250, 0.5)
bench_volatility = bench_daily_return.std(ddof=1) * np.power(250, 0.5)
print "策略波动率为", volatility
print "基准波动率为", bench_volatility
```

### 六、夏普比率 Sharpe

#### 描述
表示每承受一单位总风险，会产生多少的超额报酬，可以同时对策略的收益与风险进行综合考虑。

#### 公式

$$ Sharpe Ratio = \dfrac{R_p-R_f}{\sigma_p} $$
$$ R_p = 策略年化收益率 $$
$$ R_f = 无风险利率 0.04 $$
$$ \sigma_p = 策略收益波动率 $$

#### 计算策略夏普比率

```python
sharpe = (annualized_return - safe_rate) / volatility
print "策略夏普比率为", sharpe
```

### 七、信息比率

#### 描述
衡量单位超额风险带来的超额收益。

信息比率越大，说明该策略单位跟踪误差所获得的超额收益越高，因此，信息比率较大的策略的表现要优于信息比率较低的基金。合理的投资目标应该是在承担适度风险下，尽可能追求高信息比率。

#### 优矿公式
$$ Information Ratio = \dfrac{R_p - Rm}{\sigma_t} $$
$$ R_p = 策略年化收益率 $$
$$ R_m = 基准年化收益率 $$
$$ \sigma_t = 策略与基准每日收益差值的年化标准差 $$

#### 计算策略信息比率

```python
# 信息比率计算公式优矿的没怎么搞懂

# 策略与基准收益的每日差值
excess = daily_return - bench_daily_return

# 策略相比基准超额年华收益
excess_return = (excess + 1).cumprod()[-1] ** (250 / hold_days)

# 方式一 (策略每日收益 - 基准每日收益)的年化均值 / (策略每日收益 - 基准每日收益)的年化标准差
info_ratio_1 = excess_return/((excess).std(ddof=1) * (250 ** 0.5))
print "信息比率1为", info_ratio_1

# 方式二 这个没看懂，优矿计算的结果和此很接近
info_ration_2 = excess.cumsum()[-1]/hold_days/(excess).std(ddof=1) * (250 ** 0.5)
print "信息比率2为", info_ration_2
```

### 八、最大回测 Max Dropdown

#### 描述
描述策略可能出现的最糟糕的情况。

#### 公式
**优矿：**
$$ Max Dropdown = max(1 - \dfrac{P_x}{P_y})) $$ 
$$ P_x, P_y = 策略某日股票与现金的总价值 (x<y)$$

**书本**
$$ Max Dropdown = max(1 - \dfrac{P_y}{P_x})) $$ 
$$ P_x, P_y = 策略某日股票与现金的总价值 (x<y)$$

<font color="red">疑问：又遇到两个公式，按计算结果显示，优矿采用的似乎不是策略文档里面的的公式</font>

#### 计算策略最大回撤

```python
# 1. 计算每日账户金额序列
daily_money = total_money * (1+daily_total_return)
# 2. 计算每日起始日至当前日中最大账户金额序列
max_daily_money = np.maximum.accumulate(daily_money)
# 3. 计算最大回撤
# (1) 优矿公式
max_dropdown_1 = np.max((max_daily_money - daily_money)/daily_money)
print "(1)策略最大回撤为", max_dropdown_1
# (2) 书本公式
max_dropdown_2 = np.max((max_daily_money - daily_money)/max_daily_money)
print "(2)策略最大回撤为", max_dropdown_2
```
