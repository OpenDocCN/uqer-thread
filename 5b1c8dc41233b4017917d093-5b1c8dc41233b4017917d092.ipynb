```python
import numpy as np
import pandas as pd
from pandas import Series
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
from sklearn.linear_model import LinearRegression
import scipy.optimize as sco

pd.options.display.width = 2000
secid ='601857,600036'           #中国石油，招商银行
return_mat=DataAPI.MktEqudAdjGet(ticker=secid,beginDate='20130506', endDate='20180506', field=u"ticker,tradeDate,closePrice",pandas="1").pivot(index='tradeDate',columns='ticker',values='closePrice').pct_change().dropna()        # 以他们收盘价作为编制基础数据并用百分比函数进行规整百分化，用dropna函数填补空缺交易日数据                                                                                                              
means=return_mat.mean()*252
print'means:'                #两只股票平均年化收益率
print means
cov=return_mat.cov()*252
print'cov:'
print cov                    #两只股票协方差矩阵


port_returns = []            
port_variance = []           #准备抓取数组数据，为后面数据规整第27，28行做准备
for p in range(4000):
        weights = np.random.random(2)
        weights /=np.sum(weights)
        port_returns.append(np.sum(means*weights))
        port_variance.append(np.sqrt(np.dot(weights.T, np.dot(cov, weights))))
port_returns3 = np.array(port_returns)
port_variance3 = np.array(port_variance)
print '收益率、方差：'
print port_returns3
print port_variance3

plt.figure(figsize = (8,4))
print plt.scatter(port_variance3, port_returns3,color='b',marker='o')
risk_free = 0.04        #规定无风险利率为4%
sec_num = len(means)    #证券数，这里我们选了两只股票，因此为sec-num为2

def random_combination():                     #下面开始画有效市场边界，夏普比率，证券市场线？
    sec_num = len(means)
    random_rets = []
    random_stds = []
    for i in range(40000):
        weights = np.random.random(sec_num)
        weights = weights / np.sum(weights)
        ret, std, _ = statistic(weights)
        random_rets.append(ret)
        random_stds.append(std)
    return(random_rets, random_stds)
def statistic(weights, means=means, cov=cov):
    sec_num = len(means)
    weights = np.asmatrix(weights).T
    means = np.asmatrix(means).T
    cov = np.asmatrix(cov)
    ret = float(weights.T * means)
    std = float(np.sqrt(weights.T * cov * weights))
    sharpe = (ret - risk_free) / std
    return(ret, std, sharpe)
def efficiency_frontier(means=means, cov=cov):
    sec_num = len(means)
    min_ret = means.min()
    max_ret = means.max()
    rets = []
    stds = []
    for ret_level in np.linspace(min_ret, max_ret, 200):
        P = matrix(2 * cov.values)
        q = matrix(np.zeros(sec_num))
        G = matrix(-np.eye(sec_num))
        h = matrix(np.zeros(sec_num))
        A = matrix(np.array([np.ones(sec_num),means.values]))
        b = matrix([1.0,ret_level])
        solvers.options['show_progress'] = False
        sol = solvers.qp(P,q,G,h,A,b)
        rets.append(ret_level)
        stds.append(np.sqrt(sol['primal objective']))
    return(rets, stds)
def min_sharpe(weights):
    return(-statistic(weights)[2])
from cvxopt import solvers, matrix
cons = {'type':'eq', 'fun':lambda x:np.sum(x) - 1.0}
bnds = sec_num * ((0,1),)
opts = sco.minimize(min_sharpe, sec_num*[1.0/sec_num], method='SLSQP', bounds=bnds, constraints=cons)
orp_ret, orp_std, orp_sharpe = statistic(opts['x'])
rets, stds= efficiency_frontier()
random_rets, random_stds=random_combination()

fig = plt.figure(figsize=(12,8))
ax = fig.add_subplot(111)
plt.plot(0,risk_free,'k*',markersize=10)              #无风险资产
ax.plot(stds,rets)                                    #有效边界
ax.plot(orp_std,orp_ret,'r*',markersize=10)           #最优风险组合
plt.scatter(random_stds, random_rets, c=np.array(random_rets)/np.array(random_stds),color='b')  #随机组合
plt.colorbar(label='Sharpe Ratio')
plt.plot((0,orp_std),(risk_free,orp_ret),c='r')
ax.set_xlim(0,0.5)
plt.show()
```
