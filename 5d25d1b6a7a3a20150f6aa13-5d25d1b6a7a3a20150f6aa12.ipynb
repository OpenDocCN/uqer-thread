```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import datetime as dt
from CAL.PyCAL import * # CAL provides classes and functions dealing with date, time, and calendars. 

sns.set()
pd.set_option('display.width', 200)
cal = Calendar('China.SSE')
```

```python
"""
市净率=每一股的 市价 / 每一股的 净资产
市盈率(本益比（Price to Earnings ratio）)＝普通股 每一股的 市场价格 ÷ 普通股 每年的 每一股的 盈利
"""
```

```python
"""
TTM处理：在计算指标中，净利润采用TTM（过去12个月净利润）计算，数据更合理
负值：当公司净利润为负或者所有者权益为负时会照成PB、PE出现负值，结合前面介绍的指标计算公式，负值一般会非常大；处理方式是将负值赋值为0
极值：当公司某一年利润或者所有者权益特别小时，会导致PB、PE出现非常大的极值情况，采用 winsorize 处理（ winsorize 原理参见笔者前一篇帖子破解Alpha对冲策略）
"""

start_date = '20180701' # 开始日期
end_date = (dt.datetime.today() - dt.timedelta(days=1)).strftime('%Y%m%d') # 截止日期

# 交易所交易日历 , XSHG表示上海证券交易所
TradeCal = DataAPI.TradeCalGet(exchangeCD=u"XSHG",beginDate=start_date,endDate=end_date,field=u"calendarDate,isWeekEnd",pandas="1")
# TradeCal
TradeCal = TradeCal[TradeCal['isWeekEnd'] == 1]

Dates = map(lambda x: x[0:4]+x[5:7]+x[8:10], TradeCal['calendarDate'])
# Dates
PB_PE = pd.DataFrame(index=Dates, data=0, columns=['total PB','total PE','closeIndex'])
# PB_PE

for date in Dates:
    universe = set_universe('HS300', date) + set_universe('ZZ500', date)
    
    # 获取多只股票历史上某一天的因子数据
    data1 = DataAPI.MktStockFactorsOneDayGet(tradeDate=date,secID=universe[:400],field=['ticker','LCAP','PB','PE'],pandas="1").set_index('ticker')
    data2 = DataAPI.MktStockFactorsOneDayGet(tradeDate=date,secID=universe[400:],field=['ticker','LCAP','PB','PE'],pandas="1").set_index('ticker')
    #LCAP	float	对数市值（Natural logarithm of total market values）。计算方法：对数市值=市值的对数。属于价值类因子。数据更新时间：T日21点30分。
    
    total = pd.concat([data1, data2])
    length_total = total.shape[0]
    total.dropna(inplace=True)
    
    # 数据量不足80%时，不具有代表性
    if float(total.shape[0]) / length_total < 0.8:
        continue
        
    # 细节处理：负值极值处理
    total['PB'][total['PB'] < 0] = 0
    total['PE'][total['PE'] < 0] = 0
    pb = winsorize(total['PB'].to_dict())
    pe = winsorize(total['PE'].to_dict())
    total['PB'][pb.keys()] = pb.values()
    total['PE'][pe.keys()] = pe.values()
    
    # DataAPI.MktEqudGet(secID=u"000001",tradeDate=date,field=u"closePrice",pandas="1")
    
    # 权重
    total['wts'] = total['LCAP'] / sum(total['LCAP'])
    
    # 指数日行情
    MktIdxdGet1=DataAPI.MktIdxdGet(indexID='000001.ZICN',tradeDate=date,field='closeIndex',pandas="0")
    MktIdxdGet1=float(MktIdxdGet1[11:20])
    
    PB_PE.loc[date] = [sum(total['wts'] * total['PB']), sum(total['wts'] * total['PE']),MktIdxdGet1]

PB_PE[['total PB','total PE']].plot(secondary_y='total PE',figsize=(25,7))
PB_PE
```

```python
PB_PE['average PB'] = PB_PE['total PB'].mean()
PB_PE['average PE'] = PB_PE['total PE'].mean()

# 绘图
PB_PE[['total PB','average PB','closeIndex']].plot(secondary_y='closeIndex',figsize=(25,7))
PB_PE[['total PE','average PE','closeIndex']].plot(secondary_y='closeIndex',figsize=(25,7))
plt.show()

PB_PE
```
