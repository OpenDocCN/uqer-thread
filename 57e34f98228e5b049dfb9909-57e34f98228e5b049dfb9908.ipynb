```python
import pandas as pd
import numpy as np

# 归一化
def normalization(series):
    return (series - min(series))/(max(series) - min(series)) 
```

```python
data=pd.read_csv('hs300.csv')
del data['Unnamed: 0']
data = data.dropna()
```

```python

data['Yield'] = 100*normalization(data['Yield'])  #防止回归系数太小，太小的话，LASSO回归参数过度稀疏
data['NetProfitGrowRate'] = normalization(data['NetProfitGrowRate'])
data['AccountsPayablesTRate'] = normalization(data['AccountsPayablesTRate'])
data['MoneyFlow20'] = normalization(data['MoneyFlow20'])
data['FiftyTwoWeekHigh'] = normalization(data['FiftyTwoWeekHigh'])
data['VR'] = normalization(data['VR'])
data['MTM'] = normalization(data['MTM'])
data['CurrentRatio'] = normalization(data['CurrentRatio'])
data['DilutedEPS'] = normalization(data['DilutedEPS'])
data['EPS'] = normalization(data['EPS'])
data['EquityToAsset'] = normalization(data['EquityToAsset'])
data['FixAssetRatio'] = normalization(data['FixAssetRatio'])
data['GrossIncomeRatio'] = normalization(data['GrossIncomeRatio'])
data['LFLO'] = normalization(data['LFLO'])
data['MA10'] = normalization(data['MA10'])
data['MA120'] = normalization(data['MA120'])
data['MA20'] = normalization(data['MA20'])
data['MA5'] = normalization(data['MA5'])
data['MA60'] = normalization(data['MA60'])
data['PB'] = normalization(data['PB'])
data['PE'] = normalization(data['PE'])
data['ROA'] = normalization(data['ROA'])
data['ROE'] = normalization(data['ROE'])
data['VOL10'] = normalization(data['VOL10'])
data['VOL120'] = normalization(data['VOL120'])
data['VOL240'] = normalization(data['VOL240'])
data['VOL20'] = normalization(data['VOL20'])
data['VOL5'] = normalization(data['VOL5'])
data['VOL60'] = normalization(data['VOL60'])
data['Hurst'] = normalization(data['Hurst'])
data['DVRAT'] = normalization(data['DVRAT'])
data['BLEV'] = normalization(data['BLEV'])
data['HBETA'] = normalization(data['HBETA'])
data['HSIGMA'] = normalization(data['HSIGMA'])
data['Skewness'] = normalization(data['Skewness'])
```

在使用 LASSO 和 Elastic Net 方法时,需要对惩 罚项的系数 λ 和 α 进行合理的选定,对 λ 的选定可依据广义交叉验证最小化的方式来确定,α 的选定依据训练样本均方误差和预测偏差均方误差最小的准则进行选定。（参考fyiqi的寻优方法，但运行时间长，而且结果不一定满意）

```python
clf2 = linear_model.Lasso(alpha=0.005)
clf2.fit(train_data,train_y)
weights_lasso = clf2.coef_
weights_lasso
```

```python
#leastic net 
from sklearn import linear_model
clf1 = linear_model.ElasticNet(alpha=0.005, l1_ratio=0.5, fit_intercept=True, 
   normalize=False, precompute=False, max_iter=1000, copy_X=True, tol=0.0001, 
   warm_start=False, positive=False, random_state=None, selection='cyclic')
clf1.fit(train_data,train_y)
weights_elastic_net = clf1.coef_
print weights_elastic_net

```

可见，LASSO 方法用模型回归系数的 L1 范数作为 惩罚来压缩模型回归系数, 使绝对值较小的系数压缩为 0, 从而实现显著性变量选择和对应参数的估￼￼￼计。 然而由于凸优化的性质,容易出现过拟合问题。
而岭回归（Ridge)可以理解为在最小化残差平方和的同时对回归系数增加一个惩罚项,它和Lasso的区别在于Lasso方法增加的是Ll惩
罚项,而形dge方法增加的则是L2惩罚。
Zou和Hastie(2005)提出了一种称为ElasticNet的变量选择方法, 通过采用在 L1 范数和 L2 范数相结合的方式, 并增加惩罚项 α, 使得 Ridge 和 LASSO 优点相结合,当a=o时,ElasticNet变成简单的岭回归;当a=l时,ElastieNet变成Lasso回归,
以高预测精确度选择稀疏模型,同时解 决群组效应问题。（不知道怎么插入图片或公式，有兴趣的可以去读相关文献）
