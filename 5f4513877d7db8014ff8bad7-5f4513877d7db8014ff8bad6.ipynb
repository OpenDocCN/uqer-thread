# 基金定投日期选择

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
from CAL.PyCAL import *
```

```python
def index_analyze(ticker):
    # 获取上证50指数的信息
    df = DataAPI.MktIdxdGet(ticker=ticker,beginDate=u"20180101",endDate=u"20200825",
                        exchangeCD=u"XSHE,XSHG",field=u"ticker,tradeDate,CHGPct",pandas="1")
    
    # 确定交易日是周几
    df['weekinf'] = df['tradeDate'].map(lambda x:pd.to_datetime(x).weekday()+1)
    df['year'] = pd.to_datetime(df['tradeDate']).dt.year
    
    # 统计周一到周五的平均涨跌幅
    df_temp = df.groupby(['year','weekinf']).mean()*100
    # print(df_temp)
    
    # 可视化
    fig = plt.figure()
    ax1 = fig.add_subplot(311)
    ax1.bar(np.arange(1,6),list(df_temp['CHGPct'][:5]),width=0.5)
    plt.xticks(np.arange(1,6)+0.25,('Mon','Tues','Wed','Thur','Fri'))
    plt.ylabel(u'平均涨幅%',fontproperties=font)

    ax2 = fig.add_subplot(312)
    ax2.bar(np.arange(1,6),list(df_temp['CHGPct'][5:10]),width=0.5)
    plt.xticks(np.arange(1,6)+0.25,('Mon','Tues','Wed','Thur','Fri'))
    plt.ylabel(u'平均涨幅%',fontproperties=font)

    ax3 = fig.add_subplot(313)
    ax3.bar(np.arange(1,6),list(df_temp['CHGPct'][10:15]),width=0.5)
    plt.xticks(np.arange(1,6)+0.25,('Mon','Tues','Wed','Thur','Fri'))
    plt.ylabel(u'平均涨幅%',fontproperties=font)
    
    plt.xlabel(ticker)
    plt.tight_layout()
    plt.savefig('./risk/'+ticker+'.png',dpi=80)
    plt.show()

    return df_temp
```

```python
# 统计沪深300、上证50、中证500、创业板指周一到周五的平均收益
df_return = pd.DataFrame()

for ticker in ['000300','000016','000905','399006']:
    index_return = index_analyze(ticker)
    df_return = pd.concat([df_return,index_return['CHGPct']],axis=1)

df_return.columns = ['沪深300','上证50','中证500','创业板']
df_return.to_excel('./risk/基金定投统计.xlsx')
df_return
```

# -结论
-	几个指数往往在周一和周二表现较好
-	从上述的结果中可以发现，在2020年以前，“黑色星期四”不是白叫的，上述几个指数的统计结果表明星期四往往是跌幅最大的
-	在2020年，几个指数的统计结果显示，周五的指数跌幅往往是最大的，这可能是因为市场中“黑天鹅”事件较多，资金担心周末出现不确定因素避险导致，那么今年可以适当改变定投日期
