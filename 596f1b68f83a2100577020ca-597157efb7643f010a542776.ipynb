# 1. 怎么获取行情
---
这个问题是最常问的，以前quartz2中有get_symbol_history，现在我们进行了统一，更加快捷高效，但是可能很多人不是很清楚返回的结构。我们举一个简单的例子。

## `例一`

我们查询了当前股票池所有股票的开盘价，最高价，最低价，并存到price这个变量里面。这下很多人就蒙了，这个Price的结构是啥，怎么调取里存储的数据？首先看看这个函数的用法。
* symbol:查询的股票，如果是'all'，则返回所有股票，注意，还包含我们的benchmark的行情。
* attribute:如果是单个属性的话，输入类似于'openPrice'这样的一个字符串就可以了，多个属性则像我这样用一个列表传入。
* time_range:查询的时间长度。
* freq:行情的频率，这里是以一天为单位的。
* style:这个我们等下再说。
* rtype:一般我们需要的都是pandas的dataframe，所以写'frame'即可，也可以不指定

然后，我们再来看看price的结构

```python
print type(price)
```

纳尼，怎么是一个dict？不应该是个dataframe吗？肯定有这样的疑问。其实这时候返回的确实是一个字典，这个字典的键为资产的代码，值为我们查询的行情比如：

```python
price['000001.XSHE']
```

想必看到这里大家应该明白了，这个字典的值才是DataFrame。如果想获取000001这支票前五天的最低价，那么一个完整的写法为：

```python
price['000001.XSHE']['lowPrice']
```

注意，这里得到的还是一个Series，像有些同学想用talib计算技术指标的，还要转换成array。是不是很麻烦，当然我们有更简单的做法。

## `例二`

```python
print type(price)
print price.keys()
```

还是一个字典，但是这时候字典的键变化了，变成了attribute，这时候，想必大家都明白了，这是因为我在把style修改成'ast'，如果查询类型是frame,那么这里a代表返回的字典，键为a，值为dataframe，dataframe中，列为symbol(s)，index为time(t)。还是很灵活的，我个人觉得有些多余。目前style就支持这两种类型。

## `例三`

```python
print type(price)
```

现在我们不要返回的值为frame了，我把它换成array，看看有什么效果。

```python
price['000001.XSHE']
```

发现了，原来这时候我们存储行情的变量price是一个嵌套的字典。也就是说price[symbol]还是一个字典，这个字典的各个键是attribute，值为array类型的，简单明了，这样就很方便了，我们用talib算一些技术指标时都不用转换了。

```python
import talib
talib.MA(price['000001.XSHE']['highPrice'], 4)
```

# 2. DynamicUniverse
---
这个动态股票池，也让一些人很纠结啊。它和set_universe有什么区别？熟悉优矿的人可能知道，它还可以在代码模式下使用。假设我们要获取沪深300指数在2017年6月16日的股票池，来查询行情。

```python
universe1 = set_universe('HS300', '20170616')
universe2 = DynamicUniverse('HS300')
price1 = DataAPI.MktEqudGet(tradeDate=u"20150513",secID=universe1, isOpen="1",field=u"closePrice,tradeDate,secID",pandas="1")
price2 = DataAPI.MktEqudGet(tradeDate=u"20150513",secID=universe2, isOpen="1",field=u"closePrice,tradeDate,secID",pandas="1")
```

现在最后一句报错了，原因就在于universe2这个时候是一个Universe类。你这样直接去查是不对的。解决方法：

```python
price2 = DataAPI.MktEqudGet(tradeDate=u"20150513",secID=universe2.preview('20170616', skip_halted=True), isOpen="1",field=u"closePrice,tradeDate,secID",pandas="1")
```

很强大，还能去除停牌。

# 5. benchmark
---
今天很多人问我benchmark怎么设置，我们先看下文档的说明。总共可以支持三种方法：
1. 五个常用指数：SHCI（上证综指）、SH50（上证50）、SH180（上证180）、HS300（沪深300）和ZZ500（中证500），可以通过其简称来定义
2. 其他指数的**secID**，例如：benchmark = '399006.ZICN'  # 策略参考标准为创业板指
3. 个股的secID，例如：benchmark = '000001.XSHE'  # 策略参考标准为平安银行
---

# 8. 怎么使用自己的股票池，怎么使用自己指定的交易时间。
---
有的用户想使用自己的动态股票池，以及自己指定的交易时间。比如，我就想在每个月15号进行交易。那么这个需求似乎refresh_rate就不满足了。怎么解决？
**准备一个DataFrame，列为股票池的并集，index为自己指定的交易时间，refresh_rate设成1，universe设成DynamicUniverse('A')**，然后在handle_data里面判断previous_date是否在这个dataframe的索引中，在，把那一列取出来，把nan去掉，再拿那一列，就是您的股票池，和context.universe取个交集。
