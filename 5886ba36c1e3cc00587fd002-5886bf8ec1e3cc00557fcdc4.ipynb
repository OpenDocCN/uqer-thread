设立获取被低估的股票的函数，数据进行了极值化、中性化、标准化处理。

```python
def GetStockList(universe,cdate):
    #时间设置，选择前天数据进行回归，用昨天的数据表明是否被低估，这里的today也就是cdate实际为account.previouse_date
    today = Date.fromDateTime(cdate)  # 向前移动一个工作日
    yesterday = cal.advanceDate(today, period)
    yesterday = yesterday.toDateTime()
    #
    allmarktValue=DataAPI.MktEqudGet(tradeDate=yesterday,secID=universe,field=u"secID,marketValue",pandas="1").set_index('secID')
    data1 = DataAPI.MktStockFactorsOneDayGet(tradeDate=yesterday,secID=universe,field='secID,ETOP,PB,DebtEquityRatio').set_index('secID')  
    #l合并市值和数据表
    datall0=pd.concat([allmarktValue,data1],axis=1)
    datall0['Bit']=datall0.marketValue/datall0.PB
    datall0['Ni']=datall0.marketValue*datall0.ETOP
    datall0['D']=datall0.Ni/1000000000000
    datall0.marketValue=datall0.marketValue.apply(lambda x:mt.log(x,10))
    datall0=datall0[datall0.Bit>0]#除去空值、0值
    datall0.Bit=datall0.Bit.apply(lambda x:mt.log(x,10))
    datall0.Ni=datall0.Ni.apply(lambda x:abs(x))
    datall=datall0[datall0.Ni>0]
    datall.Ni=datall.Ni.apply(lambda x:mt.log(abs(x),10))
    datall['DNi']=datall.D.apply(lambda x:int(float(1)+x))*datall.Ni
    #市值
    marketV1=standardize(neutralize(winsorize(datall.marketValue.dropna().to_dict()),yesterday.strftime('%Y%m%d')))
    LnBit1=standardize(neutralize(winsorize(datall.Bit.dropna().to_dict()),yesterday.strftime('%Y%m%d')))
    LnDni1=standardize(neutralize(winsorize(datall.DNi.dropna().to_dict()),yesterday.strftime('%Y%m%d')))
    Lev1=standardize(neutralize(winsorize(datall.DebtEquityRatio.dropna().to_dict()),yesterday.strftime('%Y%m%d')))
    #字典是没有顺序的，要重新构建组合
    Totalt1 = pd.DataFrame(index=datall.index, columns=['marketV1','LnBit1','LnDni1','Lev1'], data=0)
    Totalt1['marketV1'][marketV1.keys()]=marketV1.values()
    Totalt1['LnBit1'][LnBit1.keys()]=LnBit1.values()
    Totalt1['LnDni1'][LnDni1.keys()]=LnDni1.values()
    Totalt1['Lev1'][Lev1.keys()]=Lev1.values()
    
    x=np.zeros((3,len(Totalt1)))
    x[0]=Totalt1.LnBit1
    x[1]=Totalt1.LnDni1
    x[2]=Totalt1.Lev1
    x=np.mat(x).T
    y=Totalt1.marketV1
    y=np.mat(y).T
    regr = linear_model.LinearRegression()
    regr.fit(x,y)

#预测#=========================================
    allmarktValue2=DataAPI.MktEqudGet(tradeDate=today,secID=universe,field=u"secID,marketValue",pandas="1").set_index('secID')
    data2 = DataAPI.MktStockFactorsOneDayGet(tradeDate=today,secID=universe,field='secID,ETOP,PB,DebtEquityRatio').set_index('secID') 

    datall20=pd.concat([allmarktValue2,data2],axis=1)
    datall20['Bit']=datall20.marketValue/datall20.PB
    datall20['Ni']=datall20.marketValue*datall20.ETOP
    datall20['D']=datall20.Ni/1000000000000
    datall20.marketValue=datall20.marketValue.apply(lambda x:mt.log(x,10))
    datall20=datall20[datall20.Bit>0]
    datall20.Bit=datall20.Bit.apply(lambda x:mt.log(x,10))
    datall20.Ni=datall20.Ni.apply(lambda x:abs(x))
    datall2=datall20[datall20.Ni>0]
    datall2.Ni=datall2.Ni.apply(lambda x:mt.log(abs(x),10))
    datall2['DNi']=datall2.D.apply(lambda x:int(float(1)+x))*datall2.Ni

    marketV2=standardize(neutralize(winsorize(datall2.marketValue.dropna().to_dict()),yesterday.strftime('%Y%m%d')))
    LnBit2=standardize(neutralize(winsorize(datall2.Bit.dropna().to_dict()),yesterday.strftime('%Y%m%d')))
    LnDni2=standardize(neutralize(winsorize(datall2.DNi.dropna().to_dict()),yesterday.strftime('%Y%m%d')))
    Lev2=standardize(neutralize(winsorize(datall2.DebtEquityRatio.dropna().to_dict()),yesterday.strftime('%Y%m%d')))
    #字典是没有顺序的，要重新构建组合
    Totalt2= pd.DataFrame(index=datall2.index, columns=['marketV2','LnBit2','LnDni2','Lev2'], data=0)
    Totalt2['marketV2'][marketV2.keys()]=marketV2.values()
    Totalt2['LnBit2'][LnBit2.keys()]=LnBit2.values()
    Totalt2['LnDni2'][LnDni2.keys()]=LnDni2.values()
    Totalt2['Lev2'][Lev2.keys()]=Lev2.values()

    x2=np.zeros((3,len(Totalt2)))
    x2[0]=Totalt2.LnBit2
    x2[1]=Totalt2.LnDni2
    x2[2]=Totalt2.Lev2
    x2=np.mat(x2).T
    #预测值
    Y=regr.predict(x2)
   
    #通过dataframe计算被低估股票并买入
    Total3= pd.DataFrame(index=datall2.index, columns=['Y','S','YS'], data=0)
    Total3['Y']=Y
    Total3['S']=Totalt2.marketV2
    Total3['YS']=Total3.Y-Total3.S
    Total3px=Total3.sort('YS',ascending=0)#0从大到小
    Buylist=Total3px[0:20].index.tolist()
    #获取买入列表========================
    #Buylist=Total3[Total3.YS>0].index.tolist()
    return Buylist
```

为了提高回测效率，本文采用了月度调仓，每月第一个交易日调仓，这里佣金设为0，是由于每次调仓都先卖出全部股票再买入，实质上操作没这么大的成交量，读者可以根据自己思路更改交易成本
诚然对于估值回归实质上过于频繁的调仓不利于估值的回归。
