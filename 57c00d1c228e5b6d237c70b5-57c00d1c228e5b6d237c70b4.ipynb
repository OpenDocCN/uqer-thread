# `【Section 2 | 分钟线模板说明】`
---
=======================================================================================================================
> * ## 期货的分钟线交易非常吸引人，在于它可以利用各种频率、各种长度的数据产生信号，在于它可以在日内进行多次操作，在于它可以做多可以做空。研究期货交易逻辑重点在于：开仓、平仓、止盈、止损。这里为大家分享一个做单品种期货的分钟线策略模板，可以很方便的完成对于单品种在某种策略逻辑的回测。


# `【增量处理数据】`
---
=======================================================================================================================
> * ## 模板基于增量处理数据的编程模式，举个例子：假设 refresh_rate 设为 r，每次 handle_date 需要长度为 2*r 历史数据进行处理和信号计算
>   * ## 正常处理数据模式
>    ![](http://g.hiphotos.baidu.com/image/pic/item/adaf2edda3cc7cd9833962653101213fb80e9171.jpg)
>      * ## 每次 get_symbol_history 都要取 2×r 长度的数据，若希望生成滚动的两根以 r 为周期的 Bar 线，每个时刻 t 都需要计算两次
>   * ## 增量处理数据模式  
>    ![](http://g.hiphotos.baidu.com/image/pic/item/728da9773912b31b2e389b958e18367adab4e132.jpg)
>      * ## 每次 get_symbol_history 只需取 r 长度的数据，若希望生成滚动的两根以 r 为周期的 Bar 线，每个时刻 t 只需计算一次，并用一个滚动寄存器 pipeline (FIFO) 不断更新
>   * ## 增量处理数据的好处
>    * ## 每次最多只需要取 refresh_rate 长度的历史数据，大大减轻了内存的压力
>    * ## 每次对于历史数据的第一级处理只需要1次，大大减少了重复计算，加快运行的速度
>    * ## 计算1个信号需要的历史数据量越多，增量处理的优势越明显


# `【模板流程】`
---
=======================================================================================================================
> * ## 模板的流程是按照如下的 4 个部分依次进行：
>   ![](http://c.hiphotos.baidu.com/image/pic/item/1e30e924b899a901e606058a15950a7b0208f55f.jpg)
>   * ##  increment 和 pipeline 用来处理数据，分别对应 “增量数据处理” 和 “数据寄存”
>   * ##  strategy 和 trader 用来写策略和交易，两者通过一个全局变量  account.signal 这个信号模板进行连接
>    * ## account.signal模板结构 pandas.DateFrame({'long': True or False,'short': True or False,'stopLoss': True or False,'stopWin': True or False}, index = [future])
>    * ## 上述信号模板可以表示 future 这个期货品种的状态，共有五种，分别对应如下
>     * ## 开多：{'long':True, 'short':False, 'stopLoss':False, 'stopWin': False}
>     * ## 开空：{'long':False, 'short':True, 'stopLoss':False, 'stopWin': False}
>     * ## 止损：{'long':False, 'short':False, 'stopLoss':True, 'stopWin': False}
>     * ## 止盈：{'long':False, 'short':False, 'stopLoss':False, 'stopWin': True}
>     * ## 空仓：{'long':False, 'short':False, 'stopLoss':False, 'stopWin': False}
>   * ##  每次 strategy 更新上述 account.signal 中期货的状态，然后 trader 根据更新后的 account.signal 进行下单交易操作
>   * ##  increment、pipeline 和 trader 已经封装好，因此只要交易模式相同，只需要修改 strategy 中的信号更新逻辑即可。


# `【Section 3 | 分钟线模板代码】`
---
=======================================================================================================================
> * ## 接下来是分钟线模板的实现代码，其适用的交易类型
>  * ## 单品种合约的分钟线交易
>  * ## 单笔交易、不加仓（ps：做期货不乱加仓是好习惯）
>  * ## 止盈、止损通过 bar 数据而非 tick 数据


```python
#*****************************************************#
#  author: Myron 
#*****************************************************#
import re
import DataAPI
import numpy as np 
import scipy as sp
import pandas as pd
import datetime as dt
import itertools as it
from collections import deque

import quartz_futures as qf
from quartz_futures.api import *
from quartz_futures.trade.cost import Slippage, Commission

import matplotlib.pyplot as plt 
from matplotlib.font_manager import FontProperties
from matplotlib.ticker import MultipleLocator,FormatStrFormatter 
```

# `【increment 子类】`
---
=======================================================================================================================
> * ## Parameters
>  * ## account: account 类对象
> * ---
> * Methods
>  * ## get_symbol_history（symbol, field, time_range）: 原生的 get_symbol_history 函数
>  * ## get_bar(barLength，**args): 返回值为一个以 symbol 为索引、field 为列的 pandas.DataFrame
>     * ## barLength: 想要得到 Bar 线的周期，通常应设置为 refresh_rate，默认值为5


```python
class increment(object):
    
    def __init__(self, futures_account):
        self.account = futures_account
        
    def get_symbol_history(self, **args):
        return self.account.get_symbol_history(**args)
        
    def get_bar(self, barLength = None, **args):
        if not barLength: barLength = 5
        data = self.get_symbol_history(**args)
        barsFrame = pd.DataFrame()
        for future in data.keys():            
            bar = {
                'closePrice': round(data[future].tail(barLength).closePrice.tolist()[-1],2),
                'openPrice': round(data[future].tail(barLength).openPrice.tolist()[0],2),
                'highPrice': round(max(data[future].tail(barLength).highPrice.tolist()),2),
                'lowPrice': round(min(data[future].tail(barLength).lowPrice.tolist()),2),
                'volume': round(sum(data[future].tail(barLength).volume.tolist()),2)
            }
            barsFrame = pd.concat([barsFrame,pd.DataFrame(bar,index = [future])])
        return barsFrame
```

# `【pipeline 子类】`
---
=======================================================================================================================
> * ## Parameters
>  * ## account: account 类对象
>  * ## barlineLength: optional，用来设置寄存器的长度
> * ---
> * Methods
>  * ## barline_update(barLength，bars): FIFO更新，pop 出最老的一根 Bar 线，append 输入的 bar 线


```python
class pipeline(object):
    
    def __init__(self, futures_account, **args):
        self.account = futures_account
        self.account.barline = {future:deque([], maxlen = args.get('barlineLength',15)) for future in self.account.universe}
        pass
    
    def barline_update(self, bars):
        for future in bars.index:
            bar = bars.loc[future,]
            self.account.barline[future].append(bar)
        return self.account.barline
        pass
```

# `【trader 子类】`
---
=======================================================================================================================
> * ## Parameters
>  * ## account: account 类对象
> * ---
> * Methods
>  * ## close_position(self, future, direction): 平仓函数
>    * ## future: optional, 要平仓的合约
>    * ## direction: optional，要平仓的合约方向，若不指定，则自动检测 position 中该合约的仓位进行操作


```python
class trader(object):
    def __init__(self,futures_account):
        self.account = futures_account
        pass

    def close_position(self,future = None,direction = None):
        if not self.account.position.get(future,0): return
        if not direction: directions = ['long','short']
        if isinstance(direction, str) == True: directions = direction.split(',')
        for direction in directions:
            if self.account.position[future].get(direction + '_position', 0) > 0:
                print self.account.current_date, self.account.current_minute, 'Close {} position on : {}'.format(direction,future),self.account.position_detail[future][0].profit, '\n'
                self.account.order(future, amount = self.account.position[future][direction+'_position'], offset_flag = 'close', direction = direction)
        pass
    
    def trade_one_future(self,future = None):
        if not future: return
        if self.account.signal.loc[future, 'short']:
            self.close_position(future,'long')
            if self.account.position.get(future, 0) and self.account.position[future].get('short_position',0):
                print self.account.current_date, self.account.current_minute, 'Hold current short position! ', '\n'
            else:
                self.account.order(future, amount = 1, offset_flag = 'open', direction = 'short')
                print self.account.current_date, self.account.current_minute, 'Open short position on : {}'.format(future), '\n'
        elif self.account.signal.loc[future, 'long']:
            self.close_position(future,'short')
            if self.account.position.get(future, 0) and self.account.position[future].get('long_position',0):
                print self.account.current_date, self.account.current_minute, 'Hold current long position! ', '\n'
            else:
                self.account.order(future, amount = 1, offset_flag = 'open', direction = 'long')
                print self.account.current_date, self.account.current_minute, 'Open long position on : {}'.format(future), '\n'
        elif self.account.signal.loc[future, 'stopLoss']:
            self.close_position(future)
            self.account.signal.loc[future] = [False]*4
            print self.account.current_date, self.account.current_minute, 'Close current position by stop loss!', '\n'
        elif self.account.signal.loc[future, 'stopWin']:
            self.close_position(future)
            self.account.signal.loc[future] = [False]*4
            print self.account.current_date, self.account.current_minute, 'Close current position by stop win!',  '\n'
```

# `【Section 4 | 策略模板】`
---
=======================================================================================================================
> * ## 策略模板包括三个部分：
>   * ## 回测全局变量的设定
>   * ## strategy 子类，操作 account.signal 进行更新
>   * ## initialize 和 handle_data 主逻辑（ 只需要把各个模块组合起来）

# `【策略示例】`
---
=======================================================================================================================
> * ## 如下给一个简单的趋势策略示例
>   * ## 品种：螺纹钢 10 月合约
>   * ## 频率：15 分钟线
>   * ## 做多条件：当前价格 >= 过去一个小时的最高价
>   * ## 做空条件：当前价格 <= 过去一个小时的最高价
>   * ## 止损比例：1.5%，计算方法： 浮动盈亏*保证金比例 / 保证金
>   * ## 止盈比例：2.5%，计算方法： 浮动盈亏*保证金比例 / 保证金


```python
#########################################################################
# environment parameters
#########################################################################
benchmark = 'rb1610'                 
universe = ['rb1610']
start = pd.datetime(2016, 5, 1)   
end = pd.datetime(2016, 8, 15)   
capital_base = 1e4                  
refresh_rate = 15                   
freq = 'm'                         
margin_ratio = DataAPI.FutuGet(ticker = universe, field = ['ticker','tradeMarginRatio'], pandas = '1')
margin_rate = dict(zip(margin_ratio.ticker.tolist(), [0.01*index for index in margin_ratio.tradeMarginRatio.tolist()]))  # 获取 universe 合约中的保证金
commission = Commission(cost=0.001, unit='perValue')     # 设置手续费
slippage = Slippage(value=0.006, unit='perValue')       # 设置滑点


#########################################################################
# strategy
#########################################################################
class strategy(object):
    
    def __init__(self, futures_account):
        self.account = futures_account
        self.future = self.account.universe[0]
    
    def update_signals(self, **args):
        barline = args['barline'] if args.has_key('barline') else None
        if not barline: return
        if barline[self.future][-1]['closePrice'] >= max([bar['highPrice'] for bar in list(barline[self.future])]):
            self.account.signal.loc[self.future] = [True,False,False,False]
        elif barline[self.future][-1]['closePrice'] <= min([bar['lowPrice'] for bar in list(barline[self.future])]):
            self.account.signal.loc[self.future] = [False,True,False,False]
        if not self.account.position_detail.get(self.future, 0): return 
        lots = self.account.position_detail[self.future][0]
        if float(lots.profit)*margin_rate.get(re.sub(r'\w+', lambda x: x.group().lower(),self.future), 0)/float(lots.margin) < -1*self.account.stopLoss:
            self.account.signal.loc[self.future] = [False,False,True,False]
        if float(lots.profit)*margin_rate.get(re.sub(r'\w+', lambda x: x.group().lower(),self.future), 0)/float(lots.margin) > self.account.stopWin:
            self.account.signal.loc[self.future] = [False,False,False,True]
        pass
        
        
#########################################################################
# quartz backtest functions
#########################################################################
def initialize(futures_account):
    futures_account.signal = pd.DataFrame({'long': False, 'short': False, 'stopLoss': False, 'stopWin': False}, index = universe)
    futures_account.increment = increment(futures_account)
    futures_account.pipeline = pipeline(futures_account,barlineLength = 4)
    futures_account.strategy = strategy(futures_account)
    futures_account.trader = trader(futures_account)
    futures_account.stopLoss = 0.015
    futures_account.stopWin = 0.025
    
def handle_data(futures_account):
    futures_account.bar = futures_account.increment.get_bar(barLength = refresh_rate)
    futures_account.barline = futures_account.pipeline.barline_update(futures_account.bar)
    futures_account.strategy.update_signals(barline = futures_account.barline)
    futures_account.trader.trade_one_future(futures_account.universe[0])
    
    
bt, account = qf.backtest.backtest(universe=universe, start=start, end=end, 
                                   initialize=initialize, handle_data=handle_data, 
                                   capital_base=capital_base, refresh_rate=refresh_rate,
                                   freq=freq, margin_rate=margin_rate, commission = commission, slippage = slippage) 
    
```

# `【展示模块】`
---
=======================================================================================================================
> * ## 鉴于目前期货回测的可视化还没有做，因此封装了一个小模块用于展示策略的表现


```python
#########################################################################
# performance modular
#########################################################################
import numpy as np
import matplotlib.pyplot as plt
from __future__ import division 

class performanceAnalysis:
    
    def __init__(self, bt = None, perf = None, **args):
        self.bt = bt
        self.perf = perf
        self.benchmark = args.get('benchmark',None)
        self.capitalBase = args.get('capital_base', 1e6)
        self.startDate = args.get('start',None)
        self.endDate = args.get('end', None)
        pass
            
    def naCurve(self):
        curve = ((np.mat(self.bt.portfolio_value)/self.capitalBase)*100).tolist()[0]
        fontsize, labelsize, linewidth = 15, 15, 2

        xmajorLocator = MultipleLocator(1)  
        ymajorFormatter = FormatStrFormatter('%.2f%%')

        plt.style.use('ggplot')
        fig = plt.figure(figsize=(12,8))
        
        ax = fig.add_subplot(1, 1, 1)
        ax.xaxis.set_major_locator(xmajorLocator)
        ax.yaxis.set_major_formatter(ymajorFormatter)     
        
        ax.spines['left'].set_color('grey')
        ax.spines['right'].set_color('grey')
        ax.spines['top'].set_color('grey')
        ax.spines['bottom'].set_color('grey')   

        with plt.style.context('fivethirtyeight'):
            plt.plot(curve,'r')
            plt.plot(self._benchmark(),'k')
        plt.grid(True)
        plt.legend(['N.A.','benchmark'],loc='best')
        plt.tick_params(labelsize=0.9*labelsize)
        plt.title('N.A.',fontsize = fontsize,verticalalignment = 'bottom',horizontalalignment = 'center')
        plt.xlabel('Time',fontsize=fontsize,verticalalignment = 'top', horizontalalignment = 'center')
        plt.ylabel('Profit', fontsize = fontsize,  verticalalignment = 'bottom', horizontalalignment = 'center',rotation=90)
        pass

    def performance(self):
        keys = ['total return','profit loss ratio','max drawdown', 'hit rate', 'action ratio']
        values = [
            str(round(self.perf['total_return']*100,2))+'%',
            round(self.perf['profit_loss_ratio'],2),
            str(round(abs(self.perf['max_drawdown'])*100,2))+'%',
            str(round(self.perf['hit_ratio']*100,2))+'%',
        ]
        perfDicts = dict(zip(keys, values))
        return pd.DataFrame(perfDicts,['performance'])
        pass
    
    def _benchmark(self):
        benchmark = DataAPI.MktFutdVolGet(ticker=self.benchmark,beginDate=self.startDate,endDate=self.endDate,field=u'closePrice',pandas="1")['closePrice'].tolist()
        benchmark = [round(100.0*index,2) for index in benchmark/benchmark[0]]
        return benchmark
        pass
        
    def _listStrip(self,list):
        tmp = []
        for index in list:
            tmp += index
        return tmp

pf = performanceAnalysis(bt, account,start = start, end = end, benchmark = universe[0], capital_base = capital_base)
pf.naCurve()
pf.performance()
```
