```python

import seaborn
from matplotlib import pylab
import quartz as qz
import pandas as pd
from pandas import DataFrame,Series
import numpy as np
from quartz.api import *
from datetime import datetime
from CAL.PyCAL import *
cal = Calendar('China.SSE')
```

```python
#筛除不合格的股票
def ini_univ(account):
    univ = account.universe
    # 1. 去除停牌股票
    univ = [s for s in univ if s in account.universe]

    # 2. 去除ST股
    try:
        STlist = DataAPI.SecSTGet(secID=univ, beginDate=yesterday, endDate=yesterday, field=['secID']).tolist()
        univ = [s for s in univ if s not in STlist]
    except:
        pass

    # 3. 去除流动性差的股票
    tv = account.get_attribute_history('turnoverValue', 20)
    mtv = {sec: sum(tvs)/20. for sec, tvs in tv.iteritems()}
    univ = [s for s in univ if mtv.get(s, 0) >= 10**7]

    # 4. 去除新上市或复牌的股票
    opn = account.get_attribute_history('openPrice', 1)
    univ = [s for s in univ if not (np.isnan(opn.get(s, 0)[0]) or opn.get(s, 0)[0] == 0)]
    return univ
```

```python
#自定义因子
def ROIC(yesterday,univ):
    
    # 信号获取，自定义
    # ROIC
    
    ROIC = {}

    # 税后净营业利润（NOPAT）= （operateProfit （营业利润）- finanExp（财务费用））*(1-经营利润所得税（33%）)
    #NOPAT
    NOPAT = DataFrame()
    profit = DataFrame()
    for rpt_type in ['Q1', 'S1', 'Q3', 'A']:
        try:
            #利润表：# 税后净营业利润（NOPAT）= （operateProfit （营业利润）- finanExp（财务费用））*(1-经营利润所得税（33%）)
            tmp  = DataAPI.FdmtISGet(secID=univ, field=['secID', 'endDate', 'publishDate', 'operateProfit','finanExp'],beginDate="20070101", publishDateEnd=yesterday, reportType=rpt_type,pandas="1")
        except:
            tmp = DataFrame()
        profit = pd.concat([profit, tmp], axis=0)

    profit = profit.sort(columns=['secID', 'endDate', 'publishDate'], ascending=[True, False, False])
    profit = profit.dropna()
    profit = profit.drop_duplicates(cols='secID')
     
    tmp  =  profit[['secID','operateProfit','finanExp']]
    NOPAT = pd.concat([NOPAT, tmp], axis=1)    
    NOPAT.set_index('secID' , inplace = True)
     
    #IC：投入资本=总资产-无息负债-超额现金=股东权益+有息负债-超额现金
    #期初有息负债=期初短期借款(STBorr)+期初长期借款(LTBorr)+期初应付债券(bondPayable)+期初一年之内到期的非流动负债(NCLWithin1Y)+期初少数股东权益(minorityInt)
    #至于超额现金，这也是一个估计数字，取决于对公司经营特点的深入了解，就A股大部分公司而言，我们可以认为它们没有任何超额现金，但像茅台这样的公司则较特殊。

    
    IC = DataFrame()
    equity = DataFrame()
    for rpt_type in ['Q1', 'S1', 'Q3', 'A']:
        try:
            #资产负债表：TShEquity（所有者权益合计）、LTBorr（长期借款）、minorityInt（少数股东权益）
            tmp  = DataAPI.FdmtBSGet(secID=univ, field=['secID', 'endDate', 'publishDate', 'TShEquity','LTBorr','minorityInt'],beginDate="20070101", publishDateEnd=yesterday, reportType=rpt_type,pandas="1")
        except:
            tmp = DataFrame()
        equity = pd.concat([equity, tmp], axis=0)

    equity = equity.sort(columns=['secID', 'endDate', 'publishDate'], ascending=[True, False, False])
    equity = equity.dropna()
    equity = equity.drop_duplicates(cols='secID')
     
    tmp  =  equity[['secID','TShEquity','LTBorr','minorityInt']]
    IC = pd.concat([IC, tmp], axis=1)    
    IC.set_index('secID' , inplace = True)

    for s in univ:
        try :
            ROIC[s] = ((NOPAT['operateProfit'][s] - NOPAT['finanExp'][s]) * (1-0.33)) / (IC['TShEquity'][s] + IC['LTBorr'][s] + IC['minorityInt'][s])       
        except:
            pass

    #引入 Series 
    ROIC = pd.Series(ROIC)
    ROIC.sort(ascending = False)
    ROIC.dropna(inplace = True)
    return ROIC
```
