```python
import pandas as pd
import numpy as np
from CAL.PyCAL import *
import matplotlib.pyplot as plt
from matplotlib import rc
rc("mathtext", default="regular")
import seaborn as sns
sns.set_style("white")
```

```python
def Calc_RPS(todayDate,RPS1=250,RPS2=120,RPS3=50):
    """
        计算三个周期段的RPS值。分别是250日RPS、120日RPS、50日RPS
    """
    cal = Calendar("China.SSE")
    # 需要计算的最后日期
    today = todayDate
    today_str = today.toISO().replace("-","")
    # 向前推进N个交易日
    period1 = "-"+str(RPS1)+"B"
    period2 = "-"+str(RPS2)+"B"
    period3 = "-"+str(RPS3)+"B"
    last_period1 = cal.advanceDate(today, period1, BizDayConvention.Preceding)
    last_period2 = cal.advanceDate(today, period2, BizDayConvention.Preceding)
    last_period3 = cal.advanceDate(today, period3, BizDayConvention.Preceding)
    last_day = [last_period1, last_period2, last_period3]
    # 获取所有当前上市的股票名单
    last_period1_str = last_period1.toISO()
    field = ["secID", "ticker", "listDate"]
    data = DataAPI.EquGet(secID=u"",ticker=u"",equTypeCD=u"A",listStatusCD=u"L",field=field,pandas="1")
    # 获取上市1年以上的股票名单、剔除上市1年以内的股票
    secID_list = data[data["listDate"] <= last_period1_str]["secID"].tolist()
    # 获取当前该股票列表的股价情况，然后剔除正在停牌的股票，重新更新股票名单。
    field = ["secID", "ticker", "closePrice","dealAmount","secShortName"]
    data_today = DataAPI.MktEqudGet(secID=secID_list,ticker=u"",tradeDate=today_str,beginDate=u"",endDate=u"",isOpen="",field=field,pandas="1")
    data_today = data_today[data_today["dealAmount"] > 0]
    secID_list = data_today["secID"].tolist()
    # 构造dataframe放置计算结果
    data_price = pd.DataFrame(index=secID_list, columns=["secShortName","price_250", "price_120","price_50","today_price", "raise_250","raise_120","raise_50","RPS_250","RPS_120","RPS_50"])
    data_price["secShortName"] = data_today["secShortName"].values
    data_price["today_price"] = data_today["closePrice"].values
    # 计算RPS的因子
    RPS_factor = 100.0/len(data_price)
    # 需要用到的str格式的日期
    last_period1_str = last_period1.toISO().replace("-","")
    last_period2_str = last_period2.toISO().replace("-","")
    last_period3_str = last_period3.toISO().replace("-","")
    # 分别获取三个日期的价格数据
    data_last_period1 = DataAPI.MktEqudAdjGet(secID=secID_list,ticker=u"",tradeDate=last_period1_str,beginDate=u"",endDate=u"",isOpen="",field=field,pandas="1")
    data_last_period2 = DataAPI.MktEqudAdjGet(secID=secID_list,ticker=u"",tradeDate=last_period2_str,beginDate=u"",endDate=u"",isOpen="",field=field,pandas="1")
    data_last_period3 = DataAPI.MktEqudAdjGet(secID=secID_list,ticker=u"",tradeDate=last_period3_str,beginDate=u"",endDate=u"",isOpen="",field=field,pandas="1")
    # 将价格数据填入
    data_price["price_250"] = data_last_period1["closePrice"].values
    data_price["price_120"] = data_last_period2["closePrice"].values
    data_price["price_50"] = data_last_period3["closePrice"].values
    # 计算三个时间段的全部个股的涨跌幅
    data_price["raise_250"] = (data_price["today_price"] - data_price["price_250"])/data_price["price_250"]
    data_price["raise_120"] = (data_price["today_price"] - data_price["price_120"])/data_price["price_120"]
    data_price["raise_50"] = (data_price["today_price"] - data_price["price_50"])/data_price["price_50"]
    # 根据计算的涨跌幅进行排序计算RPS
    data_price = data_price.sort_values(by="raise_250",ascending=False)
    data_price["RPS_250"] = data_price["raise_250"].rank()*RPS_factor
    data_price = data_price.sort_values(by="raise_120",ascending=False)
    data_price["RPS_120"] = data_price["raise_120"].rank()*RPS_factor
    data_price = data_price.sort_values(by="raise_50",ascending=False)
    data_price["RPS_50"] = data_price["raise_50"].rank()*RPS_factor
    data_price = data_price.sort_values(by="RPS_250",ascending=False)
    return data_price
```

```python
def Calc_RPS_Enter_Exit(date):
    """
        输入日期参数之后，会自动计算该日期以及之前一个交易日的三个周期RPS情况。然后分别得到新进的大于90的250RPS、120RPS、50RPS名单。
        最后返回三个dataframe、分别是新进的大于90的250RPS、120RPS、50RPS名单。
    """
    cal = Calendar("China.SSE")
    today = date
    today_str = today.toISO().replace("-","")
    # 向前推进1个交易日
    last_day = cal.advanceDate(today, "-1B", BizDayConvention.Preceding)
    last_day_str = last_day.toISO().replace("-","")
    # today_RPS
    today_RPS = Calc_RPS(today)
    # last_rps
    last_RPS = Calc_RPS(last_day)
    # 分别计算股票名单然后求差异情况
    RPS_250_last_day = last_RPS.sort_values(by="RPS_250",ascending=False)
    RPS_250_last_day = RPS_250_last_day[RPS_250_last_day["RPS_250"] >= 90].index.values.tolist()
    # 
    RPS_250_today_day = today_RPS.sort_values(by="RPS_250",ascending=False)
    RPS_250_today_day = RPS_250_today_day[RPS_250_today_day["RPS_250"] >= 90].index.values.tolist()
    # 
    RPS_250_Enter_today = set(RPS_250_today_day) - set(RPS_250_last_day)
    # 
    RPS_250_Exit_today =  set(RPS_250_last_day) - set(RPS_250_today_day) 
    #----------------------------------------------------------------------------------------------
    RPS_120_last_day = last_RPS.sort_values(by="RPS_120",ascending=False)
    RPS_120_last_day = RPS_120_last_day[RPS_120_last_day["RPS_120"] >= 90].index.values.tolist()
    # 
    RPS_120_today_day = today_RPS.sort_values(by="RPS_120",ascending=False)
    RPS_120_today_day = RPS_120_today_day[RPS_120_today_day["RPS_120"] >= 90].index.values.tolist()
    # 
    RPS_120_Enter_today = set(RPS_120_today_day) - set(RPS_120_last_day)
    # 
    RPS_120_Exit_today =  set(RPS_120_last_day) - set(RPS_120_today_day) 
    #----------------------------------------------------------------------------------------------
    RPS_50_last_day = last_RPS.sort_values(by="RPS_50",ascending=False)
    RPS_50_last_day = RPS_50_last_day[RPS_50_last_day["RPS_50"] >= 90].index.values.tolist()
    # 
    RPS_50_today_day = today_RPS.sort_values(by="RPS_50",ascending=False)
    RPS_50_today_day = RPS_50_today_day[RPS_50_today_day["RPS_50"] >= 90].index.values.tolist()
    # 
    RPS_50_Enter_today = set(RPS_50_today_day) - set(RPS_50_last_day)
    # 
    RPS_50_Exit_today =  set(RPS_50_last_day) - set(RPS_50_today_day)
    
    #-------------------------------------------------------------------------------------------------
    # 根据上面的计算情况得到的secid list来切片得到股票的dataframe
    a = []
    for x in today_RPS.index.values.tolist():
        if x not in RPS_250_Enter_today:
            a.append(False)
        else:
            a.append(True)
    RPS_250_Enter_today = today_RPS[a]
    ###
    a = []
    for x in today_RPS.index.values.tolist():
        if x not in RPS_120_Enter_today:
            a.append(False)
        else:
            a.append(True)
    RPS_120_Enter_today = today_RPS[a]
    ###
    a = []
    for x in today_RPS.index.values.tolist():
        if x not in RPS_50_Enter_today:
            a.append(False)
        else:
            a.append(True)
    RPS_50_Enter_today = today_RPS[a]
    
    return RPS_250_Enter_today, RPS_120_Enter_today, RPS_50_Enter_today
```
