# 行业动量因子分析

**来源：
20210709-广发证券-重构量化行业轮动框架：技术篇，考虑拥挤度的细分行业动量策略**

# 原理
## 动量因子
持有过去N月涨幅的M个行业（本次测试M=20）
### 阈值
研报中前10名不设置阈值，10-40名排除涨幅低于阈值的行业。阈值-0.05（跌幅大于5%）。本次未考虑这种方法，所有行业统一设置阈值。

# 实现
1. 数据：选择每月第一个交易日
1. 数据：使用DataAPI.MktInstEqudGet获取历史数据，计算各行业每月涨幅
1. 主程序
    1. 每月计算前N月涨幅最大且大于阈值的M个行业
    1. 计算当月选中的行业，取得下月收益均值
    1. 汇总


```python
from lib.functions import *
start_date='20100101'
end_date='20211001'
```

# Step1:获取所有数据

```python
# 获取申万二级行业
indexSymbol=list(DataAPI.IndustryGet(industryVersionCD=u"010303",industryLevel=u"2",
                                        isNew=1,field=u"indexSymbol",pandas="1")['indexSymbol'])
# 获取所有数据
data=DataAPI.MktIdxmGet(ticker=indexSymbol,beginDate=start_date,endDate=end_date)
# 获取期间月末交易日列表
cal=list(data.endDate.drop_duplicates().sort_values(ascending=True))
#基准收益（所有行业等权配置）
base=data[['endDate','chgPct']].dropna().groupby('endDate').mean()
base.columns=['benchmark']
```

```python
priceData=data[['ticker','endDate','closePrice']]
priceData=priceData.groupby(['ticker','endDate']).mean().unstack().T.loc['closePrice']
```

```python
# 计算所有行业前N月涨幅
def get_zf(N):
    zf=priceData/priceData.shift(N)-1
    return zf
```

# step2:动量窗口期对比

```python
# 计算收益
result=base[:]
M=20#买入排名前M的行业
for N in [1,12]:#窗口期分别为前1月、6月、12月
    rtn={}
    zf=get_zf(N)
    for i in range(12,len(cal)-1):#便于对比，统一从第12个月开始
        tmp_zf=zf.loc[cal[i]].sort_values(ascending=False).dropna()
        to_hold=list(tmp_zf[:M].index)
        if len(to_hold)>0:
            tmp_rtn=(priceData.loc[cal[i+1]]/priceData.loc[cal[i]]-1).loc[to_hold].mean()
        else:
            tmp_rtn=0
        rtn[cal[i+1]]=tmp_rtn
    stg_rtn=pd.Series(rtn)
    result[N]=stg_rtn
result.sort_index(ascending=True,inplace=True)
result.index=pd.DatetimeIndex(result.index)
```

```python
(result+1).cumprod().plot()
```

## 结论：行业动量因子能取得明显超额收益

# step3：加入阈值的对比

```python
# 计算收益
result=base[:]
M=20#买入排名前M的行业
for N in [1,12]:#窗口期分别为前1月、6月、12月
    rtn={}
    zf=get_zf(N)
    for i in range(12,len(cal)-1):#便于对比，统一从第12个月开始
        tmp_zf=zf.loc[cal[i]].sort_values(ascending=False).dropna()
        to_hold=list(tmp_zf[tmp_zf>0][:M].index)
        if len(to_hold)>0:
            tmp_rtn=(priceData.loc[cal[i+1]]/priceData.loc[cal[i]]-1).loc[to_hold].mean()
        else:
            tmp_rtn=0
        rtn[cal[i+1]]=tmp_rtn
    stg_rtn=pd.Series(rtn)
    result[N]=stg_rtn
result.sort_index(ascending=True,inplace=True)
result.index=pd.DatetimeIndex(result.index)
```

```python
(result+1).cumprod().tail(1)
```

```python
(result+1).cumprod().plot()
```

**注意如果把阈值设为-5%，则12月因子收益更高。**
# 结论：不同因子设置一定阈值收益更高。
