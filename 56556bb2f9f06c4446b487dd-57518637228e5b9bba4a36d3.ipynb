# DataFrame中加入循环求教

```python
import pandas as pd
from pandas import Series,DataFrame
import numpy as np
import seaborn as sns
sns.set_style('white')
security= pd.read_csv(u'HS300.csv')
security['tradedate']=pd.to_datetime(security['Date'])
security = security.set_index('tradedate')
window_short = 12
security['short_window'] = np.round(pd.rolling_mean(security['Close'], window=window_short), 2)
security[['Close','short_window']].plot(grid=False, figsize=(12,8))
sns.despine()
print security['Close']
#求教，代码如下，已把收盘价以index为tradedate列出，接下来想用每周的收盘价和4周前（-4）的收盘价作对比，若相减大于0，则买入，应该如何编码，是加入循环吗，还请指教，非常谢谢。
```
