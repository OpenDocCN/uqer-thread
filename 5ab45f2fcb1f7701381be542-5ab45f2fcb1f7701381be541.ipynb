## **前言：价值投资是一种信仰**

-    价值投资甚至可以用一句话概括--买入被低估的股票。然而在纷杂的信息中，判断公司是否被低估的第一部就是判断公司的价值。而格雷厄姆上面那句话清晰地表明了立场，公司的价值是现在的价值，而不是未来的价值。所以我理解的价值投资是寻找基于已知信息情况下，市场定价偏低的股票。

-    基于价值投资的理念，在去年11月份我曾经分享过一篇《ROE选股策略的思考与验证》。那篇分享的主要策略是用量化的方法寻找历史上稳定高ROE，估值较低的股票。今天的分享是基于Piotroski 提出的FSCORE体系在A股中的表现。主要参考Piotroski在2002年发表的学术论文《Value Investing: The Use of Historical Financial Statement Information to Separate Winners from Losers》。

## **一、F-SCORE 方法简介**

    Poitroski 的 F-SCORE 方法通过以下9个标准来考察公司的财务状况：
    	1、正的现金净流量（Positive net income）
    	2、正的经营性现金流（Positive operating cash flows）
    	3、正的资产增长率（Higher return on assets than the previous year）
    	4、营运现金流大于净利润（Operating cash flows greater than net income）
    	5、负的杠杆率同比增长率（Lower debt than the previous year）
    	6、流动比率同比增长率为正（Higher current ratio than previous year）
    	7、相比前一年度更低的股票稀释（Less stock dilution than the previous year）
    	8、正的毛利率同比增长率（Higer gross margin than the previous year）
    	9、正的资产周转率同比增长率（Higher asset turnover than the previous year） 
    F-SCORE 是对公司以上9条进行判断，每满足一条得一分，通过对比不同公司的得分可以得到投资价值比较高的公司。
    对F-SCORE方法进行分析，可以看到第一、二、四条在评价公司的现金流，这是一个风险控制条件，因为现金流是一个公司健康运营的保障，公司的破产很大部分是由于现金不足引起的。
    第三条是用来判断公司是在扩张还是在收缩，一般认为，由于规模效应等因素，公司规模越大，公司状况越好，未来的回报也会越大。
    第五条是在说公司的杠杆应该逐年降低，这里的假设是赚的钱优先偿还贷款，如果亏损，为了维持公司运营那么杠杆就会越来越大。
    第六条流动比率是流动资产对流动负债的比率，用来衡量企业流动资产在短期债务到期以前，可以变为现金用于偿还负债的能力。一般说来，比率越高，说明企业资产的变现能力越强，短期偿债能力亦越强；反之则弱。这里要求流动比率每年都要增加，即公司的偿债能力越来越强。
    这里不考虑第七条，第八条中毛利率反映的是一个商品经过生产转换内部系统以后增值的那一部分。也就是说，增值的越多毛利自然就越多。比如产品通过研发的差异性设计，对比竞争对手增加了一些功能，而边际价格的增加又为正值，这时毛利也就增加了。因此第八条反映的是公司的竞争能力持续增强。
    第九条中资产周转率是考察企业资产运营效率的一项很重要指标，体现企业经营期间全部资产从投入到产出的流转速度，反映企业全部资产的管理质量和利用效率。这里考察的是公司资产利用效率是不是逐年提高。

## **二、根据 F-SCORE 方法构建 F-SCORE 因子**
	
    根据以上的分析，需要在考虑将以上各个条件情况下构建可以进行对比的因子。注意到需要构建的因子是考虑将各个子因子等权相加，所以所有的子因子都要同向。
    （1）现金流：对现金流量考虑用每股经营活动产生的现金流的同比增长率和经营活动产生的现金流量净额与营业收入之比。
    经营活动产生的现金净流量与营业收入之比能反映公司销售收入中多少是以现金的形式实现的，侧面反映了公司产品的销售状况，这一项也同时能保证公司现金流为正。每股经营活动产生的现金净流量同比增长率这一指标反应了公司现金流的增长情况。
    （2）公司发展：这一项采用净资产收益率和净资产收益率的同比增长率。
    净资产收益率便于在公司中寻找质地较好的公司，因为净资产增长率反映的是每单位资产能够产生多少利润。而净资产收益率增长率则对应上面的第三条。
    （3）公司的流动性：这一项采用流动比率同比增长率，对应F-SCORE的第六条。
    （4）公司竞争力：这一指标采用销售毛利率的同比增长率，对应第八条。
    （5）资产周转率同比增长率：这一指标采用固定资产周转率同比增长率，对应于第九条。
    （6）杠杆率：
    由于需要所有的指标为正项指标，这里用股东权益比率增长率来反映企业的杠杆率的变化。企业杠杆率越低，那么股东权益比率就越高，因此增长率为正的情况下，企业的杠杆率在降低。
    而固定资产同比增长率剔除了流动资产的影响，更清晰地反映企业资产的利用效率。
    
	总结下来，这里取的八个基本因子是
    	1、ROE（净资产收益率）、
    	2、dROE（净资产收益率同比增长率）
        3、CRS（经营活动产生的现金流量净额与营业收入之比）
    	4、dCFO（每股经营活动现金流同比增长率）
    	5、dLEVER（权益比率同比增长率）、
    	6、dLIQUID（流动比率同比增长率）
    	7、dMARGIN（销售毛利率同比增长率）、
    	8、dTURN（固定资产周转率同比增长率）。

## **三、计算 F-SCORE 因子**

    下面主要是计算因子，步骤是先计算八个因子，并对每个因子进行去极值，中性化和标准化处理，然后等权加和得到FSCORE因子。

	下面写了一些在计算各个因子时需要用到的小工具。

```python
#对信号进行预处理，主要是进行去极值，标准化、中性化
def signalPre(factor,strDate):
    yesterday = getLastTradeDay(strDate)
    return standardize(neutralize(winsorize(factor), yesterday))
#获取上一交易日
def getLastTradeDay(strDate):
    cal_dates = DataAPI.TradeCalGet(exchangeCD=u"XSHG",beginDate=strDate, endDate=strDate).sort('calendarDate')
    prevTradate = cal_dates.loc[0,'prevTradeDate']
    prevTradate = datetime.datetime.strptime(prevTradate,'%Y-%m-%d').strftime('%Y%m%d')
    return prevTradate
#获取过去n个交易日的日期
def getNLastTradeDay(strDate,n):
    try:
        date = datetime.datetime.strptime(strDate,'%Y%m%d')
    except ValueError as e:
        date = datetime.datetime.strptime(strDate,'%Y-%m-%d')
    begin = (date - datetime.timedelta(n+50)).strftime('%Y%m%d')
    cal_dates = DataAPI.TradeCalGet(exchangeCD=u"XSHG",beginDate=begin, endDate=strDate).sort('calendarDate')
    if cal_dates.iloc[-1]['isOpen'] == 1:
        n = n+1
    cal_dates = cal_dates.loc[cal_dates['isOpen']==1]
    prevTradate = cal_dates.iloc[-n]['calendarDate']
    prevTradate = datetime.datetime.strptime(prevTradate,'%Y-%m-%d').strftime('%Y%m%d')
    return prevTradate
#获取未来n个交易日的日期
def getNNextTradeDay(strDate,n):
    try:
        date = datetime.datetime.strptime(strDate,'%Y%m%d')
    except ValueError as e:
        date = datetime.datetime.strptime(strDate,'%Y-%m-%d')
    end = (date + datetime.timedelta(n+50)).strftime('%Y%m%d')
    cal_dates = DataAPI.TradeCalGet(exchangeCD=u"XSHG",beginDate=strDate, endDate=end).sort('calendarDate')   
    if cal_dates.iloc[0]['isOpen'] == 1:
        n = n+1
    cal_dates = cal_dates.loc[cal_dates['isOpen']==1]
    prevTradate = cal_dates.iloc[n-1]['calendarDate']
    prevTradate = datetime.datetime.strptime(prevTradate,'%Y-%m-%d').strftime('%Y%m%d')
    return prevTradate
#获取未来n个交易日的日期
#时间类型转换，将类似'2017-01-01'形式转化为'20170101'的形式
def timeFormate(strDate):
    return datetime.datetime.strptime(strDate,'%Y-%m-%d').strftime('%Y%m%d')

#获取指定日期的前一年日期
def previousYearDate(strDate):
    try:
        date = datetime.datetime.strptime(strDate,'%Y%m%d')
    except ValueError as e:
        date = datetime.datetime.strptime(strDate,'%Y-%m-%d')
    previousYear = date.year-1
    previousMonth = date.month
    previousDay = date.day
    try:
        previousDate = datetime.datetime(previousYear,previousMonth,previousDay)
    except ValueError as e:
        previousDate = datetime.datetime(previousYear,previousMonth,previousDay-1)
    previousDate = previousDate.strftime('%Y%m%d')
    return previousDate

#如果当前日期是交易日，返回当前日期，若不是，则返回上一交易日
def getNearestTradeDay(strDate):
    cal_dates = DataAPI.TradeCalGet(exchangeCD=u"XSHG",beginDate=strDate, endDate=strDate).sort('calendarDate')
    cal_date = cal_dates.loc[0]
    if cal_date['isOpen'] == 1:
        return strDate
    else:
        return timeFormate(cal_date['prevTradeDate'])
    
#对指定因子名称的因子，进行去极值，中性化和标准化处理

def getPreFactor(begin,end,factorName):
    cal_dates = DataAPI.TradeCalGet(exchangeCD=u"XSHG",beginDate=begin, endDate=end).sort('calendarDate')
    cal_dates = cal_dates.loc[cal_dates['isOpen']==1]                                                   #取出交易日
    cal_dates['calendarDate'] = cal_dates['calendarDate'].apply(timeFormate)
    all_dates = cal_dates['calendarDate'].values.tolist()
    allFactor = DataFrame()
    for strDate in all_dates:
        factor = DataAPI.MktStockFactorsOneDayProGet(tradeDate=strDate,secID=u"",ticker=u"",field=u"secID,tradeDate,"+factorName,pandas="1")           #获取原始数据
        factor.columns = ['secID','tradeDate','value']
        factor.dropna(inplace=True)
        factor = factor.loc[(factor['value']!=np.inf) &(factor['value']!=-np.inf)] #去掉正负无穷值
        factor.set_index('secID',inplace=True)
        if len(factor) <= 1:
            continue
        factor['value'] = signalPre(factor['value'],strDate)          #对因子进行去极值，中性化，标准化
        factor = factor.reset_index()
        allFactor = allFactor.append(factor,ignore_index=True)
    return allFactor
#计算指定因子名称因子，在一定时间范围内的同比增长率

def getYoyFactor(begin,end,factorName):
    cal_dates = DataAPI.TradeCalGet(exchangeCD=u"XSHG",beginDate=begin,endDate=end).sort('calendarDate')
    cal_dates = cal_dates.loc[cal_dates['isOpen']==1]
    cal_dates['calendarDate'] = cal_dates['calendarDate'].apply(timeFormate)
    all_dates = cal_dates['calendarDate'].values.tolist()
    allDfactor = DataFrame()
    for strDate in all_dates:
        factor = DataAPI.MktStockFactorsOneDayProGet(tradeDate=strDate,secID=u"",ticker=u"",field=u"secID,tradeDate,"+factorName,pandas="1").dropna()  #取出当天的ROE
        factor.columns = ['secID','tradeDate','factor']
        previousDate = getNearestTradeDay(previousYearDate(strDate))            #得到上一年度同期交易日期
        previousFactor = DataAPI.MktStockFactorsOneDayProGet(tradeDate=previousDate,secID=u"",ticker=u"",field=u"secID,"+factorName,pandas="1").dropna()
        previousFactor.columns = ['secID','pre']
        factor = pd.merge(factor,previousFactor,on='secID',how='inner')
        factor['value'] = factor['factor']/factor['pre'] - 1                    #计算同比增长率
        factor = factor.dropna()                                                #去掉na值
        factor = factor.loc[(factor['value']!=np.inf) &(factor['value']!=-np.inf)] #去掉正负无穷值
        factor.set_index('secID',inplace=True)
        if len(factor) <= 1:
            continue
        factor['value'] = signalPre(factor['value'],strDate)                    #计算出的因子进行去极值，中性化和标准化处理
        factor.reset_index(inplace=True) 
        allDfactor = allDfactor.append(factor,ignore_index=True)
    allDfactor = allDfactor[['secID','tradeDate','value']]
    return allDfactor
```

	下面一组代码计算了八个因子，并进行等权组合得到Fscore因子。

```python
#获取指定日期范围内的进行完处理后的ROE（TTM）因子
def getRoeFactor(begin,end):
    roe = getPreFactor(begin,end,'ROE')
    roe.rename(columns={'value':'ROE'},inplace=True)
    return roe

#计算ROE增长率因子
def getDroeFactor(begin,end):
    droe = getYoyFactor(begin,end,'ROE')
    droe.rename(columns={'value':'DROE'},inplace=True)
    return droe

#获取经营活动产生的现金流量净额与营业收入之比因子
def getCrsFactor(begin,end):
    crs = getPreFactor(begin,end,'CashRateOfSales')
    crs.rename(columns={'value':'CRS'},inplace=True)
    return crs

#计算权益比率同比增长率
def getLeverFactor(begin,end):
    lever = getYoyFactor(begin,end,'EquityToAsset')
    lever.rename(columns={'value':'LEVER'},inplace=True)
    return lever

#获取指定日期的流动比率增长率因子
def getdLiquidFactor(begin,end):   
    liquid = getYoyFactor(begin,end,'CurrentRatio')
    liquid.rename(columns={'value':'DLIQUID'},inplace=True)
    return liquid

#计算每股经营活动现金流同比增长率
def getDcfoFactor(begin,end):
    dcfo = getYoyFactor(begin,end,'NetOperateCFTTM')
    dcfo.rename(columns={'value':'DCFO'},inplace=True)
    return dcfo

#计算销售毛利率同比增长率
def getDmarginFactor(begin,end):
    dmargin = getYoyFactor(begin,end,'GrossIncomeRatio')
    dmargin.rename(columns={'value':'DMARGIN'},inplace=True)
    return dmargin

#计算固定资产周转率同比增长率
def getDturnFactor(begin,end):
    dturn = getYoyFactor(begin,end,'FixedAssetsTRate')
    dturn.rename(columns={'value':'DTURN'},inplace=True)
    return dturn

#计算FSCORE因子
def getFscore(date):
    begin,end=date,date
    roe = getRoeFactor(begin,end)
    droe = getDroeFactor(begin,end)
    fscore = pd.merge(roe,droe,on=['secID','tradeDate'], how='inner')
    crs = getCrsFactor(begin,end)
    fscore = pd.merge(fscore,crs,on=['secID','tradeDate'], how='inner')
    lever = getLeverFactor(begin,end)
    fscore = pd.merge(fscore,lever,on=['secID','tradeDate'], how='inner')
    dliquid = getdLiquidFactor(begin,end)
    fscore = pd.merge(fscore,dliquid,on=['secID','tradeDate'], how='inner')
    dcfo = getDcfoFactor(begin,end)
    fscore = pd.merge(fscore,dcfo,on=['secID','tradeDate'], how='inner')
    dmargin = getDmarginFactor(begin,end)
    fscore = pd.merge(fscore,dmargin,on=['secID','tradeDate'], how='inner')
    dturn = getDturnFactor(begin,end)
    fscore = pd.merge(fscore,dturn,on=['secID','tradeDate'], how='inner')
    fscore['FSCORE'] = (fscore['ROE']+fscore['DROE']+fscore['CRS']+fscore['LEVER']+fscore['DLIQUID']+fscore['DCFO']+fscore['DMARGIN']+fscore['DTURN'])/8
    return fscore
```

	下面计算从2008年1月1日到2017年12月31日的FSCORE因子，并保存在本地。

```python
begin='20080101'
end = '20171031'
cal_dates = DataAPI.TradeCalGet(exchangeCD=u"XSHG",beginDate=begin, endDate=end).sort('calendarDate')
cal_dates['calendarDate'] = cal_dates['calendarDate'].apply(timeFormate)
cal_dates = cal_dates.loc[cal_dates['isMonthEnd']==1]
date_list = cal_dates['calendarDate'].tolist()
factor = DataFrame()
for date in date_list:
    #print(date)
    factord = getFscore(date)
    factor = factor.append(factord,ignore_index=True)
factor.to_csv('FSCORE.csv')
```

## **四、对 F-SCORE 因子进行分析**

- 接下来对因子进行系统的分析，首先需要计算股票的历史区间的收益情况、未来时间的收益情况、股票市值和换手率情况。代码如下：

```python
def getBackwardReturnsAll(universe, begin, end, window, file_name):
    #计算各股票历史区间回报率，过去window天收益率
    print str(window) + ' days backward returns will be calculated for ' + str(len(universe)) + ' stocks:'
    start_time = time.time()
    begin = getNLastTradeDay(begin,window)
    data = DataAPI.MktEqudAdjGet(secID=universe, beginDate=begin, endDate=end,field='secID,tradeDate,closePrice')            #拿取数据
    ret_data = data.groupby('secID').apply(lambda x:backwardReturns(x,window))
    ret_data = ret_data.reset_index('secID')
    ret_data.dropna(inplace=True)
    ret_data = ret_data.pivot(index='tradeDate',columns='secID',values='backwardReturns')
    print(str(np.round(time.time()-start_time,0)) + ' seconds elapsed.')
    ret_data.to_csv(file_name)
    return ret_data
    
def backwardReturns(data,window):
    data = data.sort('tradeDate')[['secID','tradeDate', 'closePrice']]
    data['backwardReturns'] = data['closePrice']/data['closePrice'].shift(window) - 1.0
    return data[['tradeDate','backwardReturns']]
```

```python
def getForwardReturnsAll(universe, begin, end, window, file_name):
    #计算各股票历史区间前瞻回报率，未来window天收益率
    end = getNNextTradeDay(end,window)
    print str(window) + ' days forward returns will be calculated for ' + str(len(universe)) + ' stocks:'
    start_time = time.time()
    data = DataAPI.MktEqudAdjGet(secID=universe, beginDate=begin, endDate=end,field='secID,tradeDate,closePrice')            #拿取数据
    ret_data = data.groupby('secID').apply(lambda x:forwardReturns(x,window))
    ret_data = ret_data.reset_index('secID')
    ret_data.dropna(inplace=True)
    ret_data = ret_data.pivot(index='tradeDate',columns='secID',values='forwardReturns')
    ret_data.to_csv(file_name)
    return ret_data
    
def forwardReturns(data,window):
    data = data.sort('tradeDate')[['secID','tradeDate', 'closePrice']]
    data['forwardReturns'] = data['closePrice'].shift(-window)/data['closePrice'] - 1.0
    return data[['tradeDate','forwardReturns']]
```

```python
def getMarketValueAll(universe, begin, end, file_name):
    # 获取股票历史每日市值
    
    print('MarketValue will be calculated for ' + str(len(universe)) + 'stocks:')
    data = DataAPI.MktEqudAdjGet(secID=universe, beginDate=begin, endDate=end,field='secID,tradeDate,marketValue')            #拿取数据
    ret_data = data.pivot(index='tradeDate',columns='secID',values='marketValue')
    ret_data.to_csv(file_name)
    return ret_data
```

```python
def getWindowMeanTurnoverRateAll(universe, begin, end, window, file_name):
    # 获取股票历史滚动窗口平均换手率
    
    print 'WindowMeanTurnoverRate will be calculated for ' + str(len(universe)) + ' stocks:'
    start_time = time.time()
    begin = getNLastTradeDay(begin,window)
    data = DataAPI.MktEqudAdjGet(secID=universe, beginDate=begin, endDate=end, field='secID,tradeDate,turnoverRate')            #拿取数据
    ret_data = data.groupby('secID').apply(lambda x:windowMeanTurnoverRateAll(x,window))
    ret_data = ret_data.reset_index('secID')
    ret_data.dropna(inplace=True)
    ret_data = ret_data.pivot(index='tradeDate',columns='secID',values='windowMeanTurnoverRate')
    ret_data.to_csv(file_name)
    return ret_data

def windowMeanTurnoverRateAll(data,window):
    data = data.sort('tradeDate')[['tradeDate','turnoverRate']]
    data['windowMeanTurnoverRate'] = pd.rolling_mean(data['turnoverRate'], window=window)
    return data[['tradeDate', 'windowMeanTurnoverRate']]
```

### **2、F-SCORE 因子对股票的预测能力分析**

	考察因子对股票的预测能力，一般是计算因子值与未来一月收益率的秩相关系数,但是考虑到基本面因子比较偏向价值投资，一般是长期持有，我们这里计算因子值与未来60个交易日（大约三个月）的秩相关系数：

```python
#读取未来60个交易日收益率数据
forward_60d_return_data = pd.read_csv('ForwardReturns_W60_FullA.csv')    # 未来60个交易日收益率 
forward_60d_return_data.set_index('tradeDate',inplace=True)
forward_60d_return_data.index = pd.to_datetime(forward_60d_return_data.index)
forward_60d_return_data.head()
```

- 计算秩相关系数

```python
#计算F-SCORE因子值与未来60个交易日（大约三个月）秩相关系数

ic_data = pd.DataFrame(index=factor_data.index, columns=['IC','pValue'])

# 计算相关系数
for dt in ic_data.index:
    tmp_factor = factor_data.ix[dt]
    tmp_ret = forward_60d_return_data.ix[dt]
    cor = pd.DataFrame(tmp_factor)
    ret = pd.DataFrame(tmp_ret)
    cor.columns = ['FSCORE']
    ret.columns = ['ret']
    cor['ret'] = ret['ret']
    cor = cor[~np.isnan(cor['FSCORE'])][~np.isnan(cor['ret'])]
    if len(cor) < 5:
        continue
    #ic,p_value = st.pearsonr(cor['FSCORE'],cor['ret'])                 # 计算相关系数   IC
    ic, p_value = st.spearmanr(cor['FSCORE'],cor['ret'])   # 计算秩相关系数 RankIC
    ic_data['IC'][dt] = ic
    ic_data['pValue'][dt] = p_value
    
# print len(ic_data['IC']), len(ic_data[ic_data.IC>0]), len(ic_data[ic_data.IC<0])
print 'IC的均值为: ', ic_data['IC'].mean()
print 'IC的标准差为：', ic_data['IC'].std()
print 'IC的中位数为： ', ic_data['IC'].median()
print 'IC的(全部, 正值, 负值)个数分别为: ', (len(ic_data), len(ic_data[ic_data.IC>0]), len(ic_data[ic_data.IC<0]))
```

- 作图

```python
fig = plt.figure(figsize=(16, 6))
ax1 = fig.add_subplot(111)
ax2 = ax1.twinx()
lns1 = ax1.plot(ic_data.index, ic_data.IC, label='IC',color='red')
lns2 = ax2.plot(ic_data.index, ic_data.pValue,'--', label='P_VALUE')

ax1.legend(['IC'],loc=2, fontsize=12)
ax2.legend(['P-VALUE'],fontsize=12)
ax1.set_ylabel(u'相关系数', fontproperties=font, fontsize=16)
ax2.set_ylabel(u'P值', fontproperties=font, fontsize=16)
ax1.set_xlabel(u'日期', fontproperties=font, fontsize=16)
ax1.set_title(u"F-SCORE因子和未来60个交易日收益之间的秩相关系数", fontproperties=font, fontsize=16)
ax1.grid()
```

### *3.1 因子的分组超额收益*
- 将因子按照因子值大小分为10组，观察不同分组的未来三个月超额收益均值。

```python
n_quantile = 10
# 统计十分位数
cols_mean = [i+1 for i in range(n_quantile)]
cols = cols_mean

excess_returns_means = pd.DataFrame(index=factor_data.index, columns=cols)

# 计算F-SCORE分组的超额收益平均值
for dt in excess_returns_means.index:
    qt_mean_results = []
    
    # 去掉nan值
    tmp_factor = factor_data.ix[dt].dropna()
    tmp_return = forward_60d_return_data.ix[dt].dropna()
    tmp_return_mean = tmp_return.mean()
    
    pct_quantiles = 1.0/n_quantile
    for i in range(n_quantile):
        down = tmp_factor.quantile(pct_quantiles*i)
        up = tmp_factor.quantile(pct_quantiles*(i+1))
        i_quantile_index = tmp_factor[(tmp_factor<=up) & (tmp_factor>=down)].index
        mean_tmp = tmp_return[i_quantile_index].mean() - tmp_return_mean
        qt_mean_results.append(mean_tmp)
        
    excess_returns_means.ix[dt] = qt_mean_results

excess_returns_means.dropna(inplace=True)
excess_returns_means.tail()
```

```python
fig = plt.figure(figsize=(12, 6))
ax1 = fig.add_subplot(111)

excess_returns_means_dist = excess_returns_means.mean()
# lns1 = ax1.plot(excess_returns_means_dist.index, excess_returns_means_dist.values, '--o', label='IC')
excess_dist_plus = excess_returns_means_dist[excess_returns_means_dist>0]
excess_dist_minus = excess_returns_means_dist[excess_returns_means_dist<0]
lns2 = ax1.bar(excess_dist_plus.index, excess_dist_plus.values, align='center', color='r', width=0.35)
lns3 = ax1.bar(excess_dist_minus.index, excess_dist_minus.values, align='center', color='g', width=0.35)
ax1.grid()
ax1.set_xlim(left=0.5, right=len(excess_returns_means_dist)+0.5)
ax1.set_ylim(-0.01, 0.004)
ax1.set_ylabel(u'超额收益', fontproperties=font, fontsize=16)
ax1.set_xlabel(u'十分位分组', fontproperties=font, fontsize=16)
ax1.set_xticks(excess_returns_means_dist.index)
ax1.set_xticklabels([int(x) for x in ax1.get_xticks()], fontproperties=font, fontsize=14)
ax1.set_yticklabels([str(x*100)+'0%' for x in ax1.get_yticks()], fontproperties=font, fontsize=14)
ax1.set_title(u"F-SCORE选股因子超额收益", fontproperties=font, fontsize=16)
ax1.grid()
```

-	我们选取市场平均收益做为基准，即简单地取市场未来60个交易日收益平均值
-  可以看到，因子在选股时六分位的超额收益最大，也并不是因子值越大，超额收益就越高。
- 因子多空收益不太明显，中间组的收益相对较好。

## **五、F-SCORE因子的历史回测**

下面来考察F-SCORE因子选股能力在历史回测中的表现：
- 回测时间段为2012年1月1日到2017年4月30日
- 股票池为全部A股
- 组合每月调仓，交易费率设为双边万分之二
- 调仓时选择股票池中F-SCORE因子最大的20%的股票

近5年回测结果表明：
  - 因子的年化收益率较高，达到26.1%，长期来看，因子选股的收益率可观，这也符合价值类因子的特征。
  - 因子的最大回撤很高，收益波动率也较大，因此该因子具有较高的风险。
  - 在回测过程中对股票采用的是等权买入方式，可以通过对选择的股票进行组合优化，来对组合的进行风险控制。

```python
# 可编辑部分与 strategy 模式一样，其余部分按本例代码编写即可

# -----------回测参数部分开始，可编辑------------
start = '2012-01-01'                       # 回测起始时间
end = '2017-04-30'                         # 回测结束时间
benchmark = 'ZZ500'                        # 策略参考标准
universe = set_universe('A')               # 证券池，支持股票和基金
capital_base = 10000000                    # 起始资金
freq = 'd'                                 # 策略类型，'d'表示日间策略使用日线回测
refresh_rate = 1                           # 调仓频率，表示执行handle_data的时间间隔

factor_data = pd.read_csv('FSCORE.csv')     # 读取量价因子数据
factor_data = factor_data[['tradeDate','secID','FSCORE']]
factor_data = factor_data.pivot(index='tradeDate',columns='secID',values='FSCORE')
factor_dates = factor_data.index.values
# ---------------回测参数部分结束----------------


# 把回测参数封装到 SimulationParameters 中，供 quick_backtest 使用
sim_params = quartz.SimulationParameters(start, end, benchmark, universe, capital_base)
# 获取回测行情数据
idxmap, data = quartz.get_daily_data(sim_params)
# 运行结果
results_corr = {}

# 调整参数(选取股票的量价因子五分位数)，进行快速回测
for quantile_five in range(1, 6):
    
    # ---------------策略逻辑部分----------------
    commission = Commission(0.0002,0.0002)      # 交易费率设为双边万分之二

    def initialize(account):                   # 初始化虚拟账户状态
        pass

    def handle_data(account):                  # 每个交易日的买入卖出指令
        pre_date = account.previous_date.strftime("%Y-%m-%d")
        if pre_date not in factor_dates:            # 只在计算过量价因子的交易日调仓
            return

        # 拿取调仓日前一个交易日的量价因子，并按照相应五分位选择股票
        pre_factor = factor_data.ix[pre_date]
        pre_factor = pre_factor.dropna()

        pre_factor_min = pre_factor.quantile((quantile_five-1)*0.2)
        pre_factor_max = pre_factor.quantile(quantile_five*0.2)
        my_univ = pre_factor[pre_factor>=pre_factor_min][pre_factor<pre_factor_max].index.values

        # 调仓逻辑
        univ = [x for x in my_univ if x in account.universe]

        # 不在股票池中的，清仓
        for stk in account.valid_secpos:
            if stk not in univ:
                order_to(stk, 0)
        # 在目标股票池中的，等权买入
        for stk in univ:
            order_pct_to(stk, 1.1/len(univ))
    # ---------------策略逻辑部分结束----------------

    # 把回测逻辑封装到 TradingStrategy 中，供 quick_backtest 使用
    strategy = quartz.TradingStrategy(initialize, handle_data)
    # 回测部分
    bt, acct = quartz.quick_backtest(sim_params, strategy, idxmap, data, refresh_rate=refresh_rate, commission=commission)

    # 对于回测的结果，可以通过 perf_parse 函数计算风险指标
    perf = quartz.perf_parse(bt, acct)

    # 保存运行结果
    tmp = {}
    tmp['bt'] = bt
    tmp['annualized_return'] = perf['annualized_return']
    tmp['volatility'] = perf['volatility']
    tmp['max_drawdown'] = perf['max_drawdown']
    tmp['alpha'] = perf['alpha']
    tmp['beta'] = perf['beta']
    tmp['sharpe'] = perf['sharpe']
    tmp['information_ratio'] = perf['information_ratio']
    
    results_corr[quantile_five] = tmp
    print str(quantile_five),
print 'done'
```

```python
fig = plt.figure(figsize=(10,8))
fig.set_tight_layout(True)
ax1 = fig.add_subplot(211)
ax2 = fig.add_subplot(212)
ax1.grid()
ax2.grid()

for qt in results_corr:
    bt = results_corr[qt]['bt']

    data = bt[[u'tradeDate',u'portfolio_value',u'benchmark_return']]
    data['portfolio_return'] = data.portfolio_value/data.portfolio_value.shift(1) - 1.0   # 总头寸每日回报率
    data['portfolio_return'].ix[0] = data['portfolio_value'].ix[0]/	10000000.0 - 1.0
    data['excess_return'] = data.portfolio_return - data.benchmark_return                 # 总头寸每日超额回报率
    data['excess'] = data.excess_return + 1.0
    data['excess'] = data.excess.cumprod()                # 总头寸对冲指数后的净值序列
    data['portfolio'] = data.portfolio_return + 1.0     
    data['portfolio'] = data.portfolio.cumprod()          # 总头寸不对冲时的净值序列
    data['benchmark'] = data.benchmark_return + 1.0
    data['benchmark'] = data.benchmark.cumprod()          # benchmark的净值序列
    results_corr[qt]['hedged_max_drawdown'] = max([1 - v/max(1, max(data['excess'][:i+1])) for i,v in enumerate(data['excess'])])  # 对冲后净值最大回撤
    results_corr[qt]['hedged_volatility'] = np.std(data['excess_return'])*np.sqrt(252)
    results_corr[qt]['hedged_annualized_return'] = (data['excess'].values[-1])**(252.0/len(data['excess'])) - 1.0
    # data[['portfolio','benchmark','excess']].plot(figsize=(12,8))
    # ax.plot(data[['portfolio','benchmark','excess']], label=str(qt))
    ax1.plot(data['tradeDate'], data[['portfolio']], label=str(qt))
    ax2.plot(data['tradeDate'], data[['excess']], label=str(qt))
    

ax1.legend(loc=0, fontsize=12)
ax2.legend(loc=0, fontsize=12)
ax1.set_ylabel(u"净值", fontproperties=font, fontsize=16)
ax2.set_ylabel(u"对冲净值", fontproperties=font, fontsize=16)
ax1.set_title(u"F-SCORE因子 - 不同五分位数分组选股净值走势", fontproperties=font, fontsize=16)
ax2.set_title(u"F-SCORE因子 - 不同五分位数分组选股对冲中证500指数后净值走势", fontproperties=font, fontsize=16)
```

- 第二张图片是对ZZ500指数做对冲的净值走势曲线。

### **2、改进思考**
* 该因子是根据 Poitroski 的 F-SCORE 方法改进得到的，由于国外判断公司的观念可能与国内不同，所以可以通过寻找财报中表现较好的价值类因子来改进此因子。
* 从因子的分组情况可以看到，因子在一分位和五分位的表现差别明显外，其他差别并不明显。可能是在构建因子时，用到了因子的具体值，而因子具体值并没有那么重要。例如在用到ROE增长率的时候，我们可以用以下逻辑在给出这个因子的暴露。如果ROE同比增长率大于零，则给1，否则给0。对每个因子都用这种方式评分，可能会改进因子对股票的区分度不显著的问题。
