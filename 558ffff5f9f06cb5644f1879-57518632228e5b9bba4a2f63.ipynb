# 通联杯2015量化投资策略大赛评分说明

通联杯2015量化投资策略大赛的评分机制说明
---------------------
***

通联杯2015量化投资策略大赛已经开始一周了，有很多参赛选手对评分有疑问，量化实验室在本帖对评分机制进行一个详细的说明，如果仍存在什么疑惑，欢迎在评论区提出来。

每个策略的大赛最终得分由两部分组成：

1、回测积分：
为了保障比赛的公平公正，在提交策略进行参赛的后，通联量化实验室将对参赛策略中的部分参数进行统一设置，包括如起止时间，初始资金，股票历史仓位，手续费标准，滑点等内容，并由系统统一进行回测，记录回测中的各个指标进行打分。
使用年化收益、年化波动率、最大回撤三个指标作为评分标准。先单独计算三个指标在参赛策略中排名，再用三个指标的排名计算策略的总体排名，若分数一样则以参赛时间为标准，参赛时间在前的选手（团队）排名在前。

2、模拟盘积分：
使用年化收益、年化波动率、最大回撤三个指标作为评分标准。先单独计算三个指标在参赛策略中排名，再用三个指标的排名计算策略的总体排名，若分数一样则以参赛时间为标准，参赛时间在前的选手（团队）排名在前。

**此外，量化实验室会根据三个指标的结果对总分有一定程度的微调，如对过低仓位策略进行降级评定等。**

回测积分占总积分的30%，模拟盘积分占总积分的70%。

需要说明的是：
参赛者在大赛排名页面可以看到此次比赛的前50名，在我的参赛策略中，可以看到参赛者提交的参赛策略的回测和模拟盘得分，以及模拟盘年化收益，模拟盘收益波动率，模拟盘最大回撤三个指标的详情。

**此次大赛于2015年6月23日开始，对于23日之前提交的策略，量化实验室只考虑其23日以后的表现（收益变化），参赛者无需担心23日之前的组合是盈是亏。**

**模拟交易中策略详情展示的是该策略整个生命周期的表现，与大赛时间无关。大赛得分依据的只是该策略在大赛期间的表现。**
