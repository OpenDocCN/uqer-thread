# 大师系列量化策略（史蒂夫路拂  罗伯特山朋 柏顿墨其尔）

价值选股法--史蒂夫路拂

市盈率为正且小于市场平均值 PE
市净率小于市场平均值      PB
市现率小于市场平均值      PCF
流动比率大于市场平均值    CurrentRatio
长期借款与资产总计之比小于市场平均值 LongDebtToAsset
股息率大于市场平均值 数据来源： EquDivGet ， 公式计算

价值型投资法--罗伯特山朋

1.  市销率为正，且低于市场平均值。 
2.  市现率为正，且低于市场平均值。
3.  公司高管持股比例大于市场平均值。改为（分析师推荐评级大于市场平均值）

成功选股法--柏顿墨其尔

1.   最近四季度市盈率最低的30%；(PE(TTM))  
2.   最近四季度净利润(TTM)增长最高的30% 
3.   连续5年内盈余成长率为正值；（最近3年净利润未出现亏损 11年之前）
4.   股票限制15只
