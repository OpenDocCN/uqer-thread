# Half_day Reversal策略

### Half_day Reversal策略的思想来自于2010年8月的一篇文章：
### 《Profitable Mean Reversion after Large Price Drops:A story of Day and Night in the S&P 500, 400 Mid Cap and 600 Small Cap Indices》，和以往的收盘价-收盘价时间段不同，文章研究了开盘价-收盘价、收盘价-开盘价期间所包含的信息和股价的反应，文中通过三个不同的策略研究了反转效应，而反转效应可能是源于投资者的过度反应或反应不足。
### 本次研究将通过我国HS300在2012-06至2014-06的数据来研究Half_day Reversal策略。
### 研究结构如下：
- ### 一、导入数据、处理数据
- ### 二、计算策略收益及画图
- ### 三、回测指标分析
- ### 四、10分位组合分析
- ### 五、总结

### 一、导入数据、处理数据

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 导入数据，加入一行NaN方便之后数据处理
clean_data = pd.read_excel("HS300 DATA.xlsx",skiprows=[0,1,2])
clean_data.loc[485]=np.NAN

# 再导入数据用于获取股票名称和代码
data = pd.read_excel("HS300 DATA.xlsx",index_col=False)
data_copy = data.copy()
data
```

#### 处理数据，方便之后的操作。

```python
# 获取含有股票名称和代码的列表
def getNameAndCode(dataframe):
    return [(dataframe.columns[i],dataframe.iloc[0][i+1]) for i in range(0,1197,4)]
    
name_and_code = getNameAndCode(data_copy)

# 最后增加一列NaN，方便数据处理
clean_data['Unnamed: 1199'] = np.NAN

# 获取每四列的数据
tmp = [clean_data.iloc[:,i:i+4] for i in range(0,1197,4)] 

# 在数据后加入名称列和代码列
for i,df in enumerate(tmp):
    df['name'] = name_and_code[i][0]
    df['code'] = name_and_code[i][1]

# 定义函数并重命名列名
def rename(df):
    df.rename(columns={df.columns[0]:'date',df.columns[1]:'open',df.columns[2]:'close',df.columns[3]:'useless'},inplace=True)
    return df
    
for i in tmp:
    i = rename(i)

# 将数据concat为一个大的DataFrame，并删除无用列
hs_data = pd.concat(tmp)
hs_data.drop('useless', axis=1, inplace=True)
```

#### 文章提及了三种策略：
- version1:买入t日收盘至t+1日开盘时表现最差的n只股票，持有至t+1日收盘时卖出；
- version2:买入t日开盘至t日收盘时表现最差的n只股票，持有至t+1日开盘时卖出；
- benchmark：买入t日收盘至t+1日收盘时表现最差的n只股票，持有至t+2日收盘时卖出。
#### 我们的股票池为HS300，文章之后用10分位来构建组合，所以这里n选择为30。
#### 股票为等权重选择。
#### 在计算收益率的时候，首先要避免未来函数，其次要找出未开盘或不在指数中以及涨跌停限制不能买卖的股票，剔除这些日数据。

```python
# 获取date的list，方便以后做索引用
date_list = list(hs_data.iloc[1:484,:]['date'])

# 判断是否开盘或是否在指数中，仅在open和close数据都存在时判定为开盘且在指数中，标记为1
hs_data['isOpen'] = np.where(np.isnan(hs_data['open']) | np.isnan(hs_data['close']), 0, 1)
# 判断是否涨停或跌停，涨跌停时标记为1
hs_data['limit'] = np.where((hs_data['close'] / hs_data['open'] > 1.1) | (hs_data['close'] / hs_data['open'] < -1.1), 1, 0)

# 求出version1、version2、benchmark的需判断的收益率和能实现的收益率，注意避免未来函数
hs_data['return_rank_v1'] = np.log(hs_data['open'].shift(-1) / hs_data['close']).shift(1)
hs_data['return_v1'] = np.log(hs_data['close'] / hs_data['open'])
hs_data['return_rank_v2'] = np.log(hs_data['close'] / hs_data['open'])
hs_data['return_v2'] = np.log(hs_data['open'].shift(-1) / hs_data['close']).shift(1)
hs_data['return_rank_v3'] = np.log(hs_data['close'].shift(-1) / hs_data['close']).shift(1)
hs_data['return_v3'] = hs_data['return_rank_v3'].shift(-1)

# 创建一列'index'为现在的index数值
hs_data['index']=hs_data.index
# 创建一列'number'为行的长度，并设置为index，为了之后处理数据方便
hs_data['number'] = range(hs_data.shape[0])
hs_data.set_index('number', inplace=True)

# 删除未开盘或不在指数中的日数据、删除涨跌停的日数据
hs_data.drop(hs_data[(hs_data['isOpen']==0)].index.tolist(),inplace=True)
hs_data.drop(hs_data[hs_data['limit']==1].index.tolist(),inplace=True)

# 将索引重新设置为'index'
hs_data.set_index('index',inplace=True)

hs_data
```

### 二、计算策略收益及画图
#### 计算version1、version2、benchmark在选择了第一个十分位组合时的收益。
#### 文章考虑了单边0.05%的交易成本，故纳入计算中。

```python
# 选出version1每天收益率最低的30只股票，计算持有这个组合的收益率
return_v1_p = [hs_data.loc[i].sort_values('return_rank_v1').iloc[:30, :]['return_v1'].mean() for i in range(1,484)]
# 新建名为return_data的DataFrame，加入一列收益率
return_data = pd.DataFrame(return_v1_p, columns=['return_v1_p'])
# 计算考虑了交易成本的每日收益率
return_data['return_v1_p_minus_cost'] = return_data['return_v1_p'] - 0.0005
# 计算考虑了交易成本的累积收益率
return_data['return_v1_p_minus_cost_cum'] = (return_data['return_v1_p_minus_cost'] + 1).cumprod()

# 同理计算version2的各项收益率
return_v2_p = [hs_data.loc[i].sort_values('return_rank_v2').iloc[:30, :]['return_v2'].mean() for i in range(1,484)]
return_data['return_v2_p'] = return_v2_p
return_data['return_v2_p_minus_cost'] = return_data['return_v2_p'] - 0.0005
return_data['return_v2_p_minus_cost_cum'] = (return_data['return_v2_p_minus_cost'] + 1).cumprod()

# 计算benchmark的各项收益率，因为策略不同，导致切片方式略有不同
return_v3_p = [0]
return_v3_p = return_v3_p + [hs_data.loc[i].sort_values('return_rank_v3').iloc[:30, :]['return_v3'].mean() for i in range(2,484)]
return_data['return_v3_p'] = return_v3_p
return_data['return_benchmark_minus_cost'] = return_data['return_v3_p'] - 0.0005
return_data['return_benchmark_minus_cost_cum'] = (return_data['return_benchmark_minus_cost'] + 1).cumprod()
return_data
```

#### 绘制version1、version2、benchmark在选择了第一个十分位组合时的累计收益图。

```python
# 设置一个日期的索引并画图
return_data['date'] = pd.to_datetime(date_list).strftime('%Y/%m/%d')
return_data.set_index('date', inplace=True)
return_data[['return_v1_p_minus_cost_cum', 'return_v2_p_minus_cost_cum', 'return_benchmark_minus_cost_cum']].plot(figsize= (22, 8), title  = 'return compare')
```

#### 可以发现，策略1的收益曲线很不错，上涨幅度很大而且回撤很小，在考虑了限制的股票及交易成本之后，依旧是很吸引人的策略；
#### 而策略2就表现平平，虽然波动和回撤很小，但收益率也很低，两年大概只有10%的收益，并不是理想的策略；
#### benchmark的表现就很差了，两年净值下跌到只剩30%多，表现很差。

### 三、回测指标分析
#### 通过回测指标来分析策略，要比图标分析更加客观，更能找到策略的具体优缺点。

```python
# 定义函数求各项回测指标
def ratio(datas):
    res = []
    for item in datas:
        data = datas[item]
        
        # 累积收益率        
        strategy_cum = (data + 1).cumprod()
        # 年化收益率
        return_year = data.mean() * 252
        # 每日收益率      
        return_avg = data.mean() 
        # 年化波动率
        volatility = data.std() * 252 ** 0.5
        # 最大每日收益
        profit_max = data.max()
        # 最大每日损失
        loss_max = data.min()
        # 信息比率               
        ir = return_year / volatility
        # 上涨天数
        num_of_up = data[data > 0].count()
        # 下跌天数
        num_of_down = data[data < 0].count()
        # 胜率
        win_rate = float(num_of_up) / (num_of_up + num_of_down)
        # 上涨时平均每日收益率
        gain_of_up = data[data > 0].mean()
        # 下跌时平均每日收益率        
        loss_of_down = data[data < 0].mean()
        # 盈亏比
        profit_loss_ratio = -(gain_of_up / loss_of_down)
        # 最大回撤
        drawdown = ((strategy_cum.cummax() - strategy_cum)/strategy_cum.cummax()).max()
        
        # 创建一个临时的DataFrame
        tmp = pd.DataFrame([ir, float((strategy_cum).tail(1)), return_year, return_avg, volatility, profit_max, loss_max,  
                            num_of_up, num_of_down, win_rate, gain_of_up, loss_of_down, profit_loss_ratio, drawdown], 
                           columns=[data.name], 
                           index=['Information Ratio', 'Cumulative Return', 'Annualised Return', 'Average return', 'Annualised Volatility', 'Maximum Daily Profit', 
                                  'Maximum Daily Loss', 'Number of Up Periods', 'Number of Down Periods', 'Win Rate', 'Avg Gain in Up Periods', 
                                  'Avg Loss in Down Periods', 'Profit and Loss Ratio', 'Maximum Drawdown'])
        res.append(tmp)
    # 返回拼接好的DataFrame
    return pd.concat(res,axis=1,join='inner')

# 输入每日收益率数据，计算各项回测指标，保留4位小数
ratio(return_data[['return_v1_p_minus_cost', 'return_v2_p_minus_cost', 'return_benchmark_minus_cost']]).round(4)
```

- 信息比率： 计算方式采用文章中 return/volatility 的方法而非 excess return/tracking error volatility，得出策略1的信息比率为2.99，远高于策略2的0.72，而benchmark收益为负故信息比率也为负；
- 累积收益： 策略1两年资产变成了3.42倍，大幅打败策略2，benchmark的净值只剩36.02%；
- 年化收益率： 策略1有66.76%的年化收益率，非常吸引人，策略2的收益率为6.46%，比一般的基金产品要好，而benchmark为-50.15%；
- 年化波动率： 策略1收益高波动也大，波动率为22.34%，相比之下策略2的波动率就很小了，只有8.98%，benchmark不仅收益率很差，甚至波动率还要大于策略1，非常不好；
- 每日最大损益： 策略1的每日最大收益为4.71%，最大亏损为5.45%，这是会造成波动率较大的原因之一，相比之下策略2的每日最大收益为6.05%，最大亏损为2.79%，波动率就相对较小，而benchmark的每日最大收益为4.99%，最大亏损为7.74%，确实benchmark的波动是最大的；
- 上涨、下跌天数： 策略1的上涨、下跌天数为（273/210），策略2和benchmark分别为（240/243）和（222/261），策略1上涨天数较多，策略2基本持平，benchmark下跌天数较多；
- 胜率： 策略1上涨天数较多，故胜率>50%，为56.5%，策略2的胜率接近50%，benchmark的胜率低于50%；
- 上涨、下跌时的平均损益： 策略1的在上涨时平均收益为1.21%，在下跌时平均亏损为0.97%，策略2和benchmark分别为（0.38%/0.32%）和（1.07%/1.28%），可以看出，策略2收益小波动小，所以每日的损益都是小于策略1和benchmark的；
- 盈亏比： 策略1的盈亏比为1.25，策略2为1.17，都是大于1的，说明上涨时的平均收益要大于下跌时的平均亏损，而benchmark的盈亏比为0.84，小于1，说明涨的少亏的多；
- 最大回撤： 策略1和策略2的最大回撤都不算大，分别为9.1%和9.6%，而benchmark几乎一直在下跌，故最大回撤也非常大。
#### 小结：
#### 策略1：信息比率高，收益高，胜率>50%，盈亏比>1，最大回撤较小，唯一的不足是波动率稍大，但依旧是很不错的策略。
#### 策略2：各项指标都比较平庸，收益小但波动率也小，适合风险厌恶者选择。
#### Benchmark：净值一直下跌，各项指标都很差，故并不是一个好的策略。

### 四、10分位组合分析
#### 将资产池等分为10个组合，研究它们的收益曲线和回测指标，分析并找到趋势。

### 策略1:买入t日收盘至t+1日开盘时表现最差的n只股票，持有至t+1日收盘时卖出。

```python
# 新建一个空的DataFrame，设置索引并加入不同分位数组合的每日收益率
v1_deciles = pd.DataFrame()
v1_deciles['index']=range(483)
v1_deciles.set_index('index')
for i in range(1, 11):
    v1_deciles['return_v1_d' + str(i)] = pd.DataFrame([hs_data.loc[j].sort_values('return_rank_v1').iloc[(i - 1) * 30 : i * 30, :]['return_v1'].mean() - 0.0005 
                                                       for j in range(1,484)])
# 设置日期索引
v1_deciles['date'] = pd.to_datetime(date_list).strftime('%Y/%m/%d')
v1_deciles.set_index('date', inplace=True)
v1_deciles.drop('index', axis=1, inplace=True)

# 画出version1的10个组合的累计收益曲线
for i in range(1,11):
    (v1_deciles['return_v1_d' + str(i)] + 1).cumprod().plot(figsize= (22, 8), title  = 'version1_portfolios')
plt.legend(loc='best')
```

### 策略1收益曲线分析：
#### 策略1前4个分位的组合能获得正收益，前2个分位的组合表现较好，这和文章中的实证结果是相吻合的；
#### 组合1的表现大幅超越其余组合，组合10则大幅低于其余组合，说明反转效应的强度和跌幅排名并非简单线性关系；
#### 10个组合的收益按分位数递减，说明策略有很好的甄别作用；

```python
# 计算version1的10个组合的各项回测指标，保留四位小数
ratio(v1_deciles[['return_v1_d'+str(i) for i in range(1,11)]]).round(4)
```

### 策略1回测指标分析：
- 信息比率： 组合1的信息比率为2.99，组合10为-2.42，10个组合按顺序递减；
- 年化收益率： 组合1有66.76%的年化收益率，组合10的收益率为-50.73%，10个组合按顺序递减；
- 年化波动率： 组合1的波动率最大，为22.34%，其余组合波动率都在20%左右；
- 每日最大损益： 10个组合的每日最大损益并无明显差距；
- 胜率： 组合1胜率最高，为56.52%，组合10的胜率最低，为43.48%，10个组合按顺序递减；
- 盈亏比： 组合1的盈亏比最高，为1.26，组合2次之，为1.20，组合10为0.88，组合3-9的盈亏比相差不大，但组合1和组合10的盈亏比差距明显；
- 最大回撤： 组合1的最大回撤最小，为9.14%，而组合10的回撤最大，为63.75%，10个组合按顺序递增。
#### 小结：
#### 组合1表现最优，组合2表现次优，组合10表现最差，其余组合的差别并不大；
#### 组合1收益最高，回撤最小，但波动率最大；
#### 组合1、2大幅优于其他组合，很可能是高胜率和高盈亏比共同作用的结果。

### 策略2：买入t日开盘至t日收盘时表现最差的n只股票，持有至t+1日开盘时卖出。

```python
# 新建一个空的DataFrame，设置索引并加入不同分位数组合的每日收益率
v2_deciles = pd.DataFrame()
v2_deciles['index']=range(483)
v2_deciles.set_index('index')
for i in range(1, 11):
    v2_deciles['return_v2_d' + str(i)] = pd.DataFrame([hs_data.loc[j].sort_values('return_rank_v2').iloc[(i - 1) * 30 : i * 30, :]['return_v2'].mean() - 0.0005
                                                       for j in range(1,484)])
# 设置日期索引
v2_deciles['date'] = pd.to_datetime(date_list).strftime('%Y/%m/%d')
v2_deciles.set_index('date', inplace=True)
v2_deciles.drop('index', axis=1, inplace=True)

# 画出version2的10个组合的累计收益曲线
for i in range(1,11):
    (v2_deciles['return_v2_d' + str(i)] + 1).cumprod().plot(figsize= (22, 8), title  = 'version2_portfolios')
plt.legend(loc='best')
```

### 策略2收益曲线分析：
#### 策略2仅第1个分位的组合获得了正的收益，其余组合均表现很差；
#### 组合1的表现大幅超越其余组合，说明反转效应的强度和跌幅排名并非简单线性关系；
#### 10个组合的收益按分位数递减，说明策略有一定的甄别作用；

```python
# 计算version2的10个组合的各项回测指标，保留四位小数
ratio(v2_deciles[['return_v2_d'+str(i) for i in range(1,11)]]).round(4)
```

### 策略2回测指标分析：
- 信息比率： 仅组合1的信息比率为正，其余组合都为负且按顺序递减；
- 年化收益率： 仅组合1有正的年化收益率，其余组合都为负且按顺序递减；
- 年化波动率： 组合1的波动率最大，为8.98%，组合10次之，为6.9%，其余组合波动率都在5.5%左右；
- 每日最大损益： 组合1有较高的每日最大损益，其余组合差距并不明显；
- 胜率： 组合1胜率最高，为49.69%，组合9的胜率最低，为25.47%，但都小于50%，且基本按顺序递减；
- 盈亏比： 仅组合1的盈亏比大于1，为1.17，其余组合均小于1，其中组合9最小，为0.58，且基本按顺序递减；
- 最大回撤： 组合1为唯一盈利的策略，最大回撤最小，为9.14%，其余组合的最大回撤按顺序递增。
#### 小结：
#### 组合1为唯一盈利的策略，其余组合均出现大幅的净值下跌；
#### 组合10并非最差的策略，组合9才是；
#### 若不考虑限制，做空组合9也可得到一个不错的策略。

### Benchmark：买入t日收盘至t+1日收盘时表现最差的n只股票，持有至t+2日收盘时卖出。

```python
# 新建一个空的DataFrame，设置索引并加入不同分位数组合的每日收益率
v3_deciles = pd.DataFrame()
v3_deciles['index']=range(483)
v3_deciles.set_index('index')
for i in range(1, 11):
    v3_deciles['return_v3_d' + str(i)] = pd.DataFrame([hs_data.loc[j].sort_values('return_rank_v3').iloc[(i - 1) * 30 : i * 30, :]['return_v3'].mean() - 0.0005 
                                                       for j in range(1,484)])
# 设置日期索引
v3_deciles['date'] = pd.to_datetime(date_list).strftime('%Y/%m/%d')
v3_deciles.set_index('date', inplace=True)
v3_deciles.drop('index', axis=1, inplace=True)

# 画出benchmark的10个组合的累计收益曲线
for i in range(1,11):
    (v3_deciles['return_v3_d' + str(i)] + 1).cumprod().plot(figsize= (22, 8), title  = 'benchmark_portfolios')
plt.legend(loc='best')
```

### Benchmark收益曲线分析：
#### benchmark仅第10个分位的组合获得了正的收益，其余组合均未取得正收益；
