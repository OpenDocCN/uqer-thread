# 放量上涨_特殊调仓止损机制

策略说明：
思路：
相对成交量=今日成交量/近5日成交量
今日涨幅=今日收盘价/今日开盘价
放量上涨：
取“相对成交量”排序前1/20（5%）的和今日涨幅排序前1/20（5%）交集为‘放量上涨“

特殊点：
持仓调整机制：优矿默认调仓为n日一个周期，每个完全卖出再买入，这样当然是最简单直观的。但这种机制其实有个缺陷，2个完全相同策略，策略开始日期相差一天，实际买卖序列可能完全不同，有偶然因素。本策略每日一刷新，但是每天只使用1/n的资金，卖出n天前的组合（理论上收回1/n资金），然后买入今天符合条件的个股。另外，单个股票持仓不超过当日可用资金的1/3（考虑到每天可用资金为1/n，所以占总资金比例1/3n)
止损：采用基准指数控制止损，如果基准指数近3天，相对最高点回撤达5%，则认为熊市，进入休眠周期，3日内只卖不买，过了休眠期，然后再继续买入。
