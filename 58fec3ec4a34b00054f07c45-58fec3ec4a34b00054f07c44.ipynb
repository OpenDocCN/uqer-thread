# 单行业股票对比

```python
import pandas as pd
industrySymbol=110101#此处定义行业Symbol
indexSymbol=850111#此处定义行业指数symbol
start_date='20100101'
end_date='20170101'
data_index=DataAPI.MktIdxmGet(beginDate=start_date,endDate=end_date,
                         ticker=str(indexSymbol),field=u"endDate,closePrice",pandas="1")
#data_index.columns=['endDate',indexSymbol]
data_index=data_index.set_index('endDate')
data_index
data_index_increase=data_index/data_index.ix[-1]
data_index_increase
```

```python
#获取行业股票
stocks=DataAPI.EquIndustryGet(industryVersionCD=u"010303",intoDate=u"20170414",field=u"",pandas="1")
stocks=stocks[stocks['industrySymbol']==industrySymbol]
stocksID=list(stocks['secID'])
data=pd.DataFrame()
for stock in stocksID:
    if len(data)==0:
        data=DataAPI.MktEqumGet(secID=stock,beginDate=start_date,endDate=end_date,
                                field=u"monthBeginDate,closePrice",pandas="1")
        data.columns=['tradeDate',stock]
    else:
        tmp=DataAPI.MktEqumGet(secID=stock,beginDate=start_date,endDate=end_date,
                               field=u"monthBeginDate,closePrice",pandas="1")
        tmp.columns=['tradeDate',stock]
        data=pd.merge(data,tmp,left_on='tradeDate',right_on='tradeDate',how='left')
data=data.set_index(pd.DatetimeIndex(data['tradeDate']))
data=data.drop('tradeDate',1) 
data_increase=data/data.ix[-1]
data_increase
```
