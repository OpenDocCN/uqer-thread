```python
# 常规地导入计算模块
import pandas as pd
import numpy as np
import datetime as dt
import matplotlib.pyplot as plt
```

- 上方是常规性导入Python模块，下方是子函数定义，用于记录每次换仓时的交易信号。细节繁琐，如果感觉没有bug，请跳过阅读，不建议深究。

```python
"""
这个Account类，用于将股票指数、标的物ETF和期权价格三种数据在每次交易信号产生后，记录换仓行为
这个Account类，作用类似于 优矿uqer回测框架中的context类
"""

class Account(object):
    def __init__(self,df_preTradeDate):
        # 换仓时间、新持仓合约、持仓数量等信息的记录总表
        self.df_preTradeDate=df_preTradeDate
        self.position_table={}
        
        # 每次换仓时，记录新持仓的期权的最后交易日，以及倒数第1第2第3第4交易日
        self.last_tradeDate=None
        self.previous_date =None
        self.lag2_date= None
        self.lag3_date= None
        self.lag4_date= None
        self.lagDays_dict={}
        

    def trade_option(self,current_date,varTicker,direction,lag_days=1):
        # 交易信号产生时，记录当天日期
        current_date0=current_date
        current_date=current_date.replace("-","")
        
        # 抽取前一天交易日，标的物的收盘价
        field="tradeDate,openPrice,preClosePrice"
        df=DataAPI.MktFunddGet(ticker=varTicker,tradeDate=current_date,field=field,pandas="1")
        closePrice = df["preClosePrice"].values[0]
        
        # 抽取交易信号当天，期权的T型报价表，并选择合适月份、合适行权价、正确Call或Put方向的期权进行交易
        field="tradeDate,secID,ticker,secShortName,lastTradeDate,strikePrice,contractType"
        df=DataAPI.OptDpoGet(beginDate=current_date,endDate=current_date,varTicker=varTicker,field=field)
        lastTradeDate_list=sorted(df["lastTradeDate"].unique())
        #print(lastTradeDate_list)
        for td in lastTradeDate_list:         
            self.last_tradeDate=td
            self.previous_date = self.df_preTradeDate.at[self.last_tradeDate,"prevTradeDate"]
            self.lag2_date= self.df_preTradeDate.at[self.previous_date,"prevTradeDate"]
            self.lag3_date= self.df_preTradeDate.at[self.lag2_date,"prevTradeDate"]
            self.lag4_date= self.df_preTradeDate.at[self.lag3_date,"prevTradeDate"]
            self.lagDays_dict={
                0:self.last_tradeDate,
                1:self.previous_date,
                2:self.lag2_date,
                3:self.lag3_date,
                4:self.lag4_date,
            }
            swap_date = self.lagDays_dict[lag_days]
            
            if current_date0<swap_date:
                df=df[ df["lastTradeDate"]==td ].reset_index(drop=True)
                break
        
        # 根据目前信号 是看多还是看空，决定新持仓哪一张期权（本策略是“卖空期权”思路）
        if direction==1:
            df=df[df["contractType"]=="PO"]
            selected_sp=min(df["strikePrice"].tolist())
            gap = selected_sp-closePrice
            for sp in df["strikePrice"].tolist():
                if sp-closePrice<0 and sp-closePrice>gap:
                    selected_sp= sp
                    gap = selected_sp-closePrice
        elif direction==-1:
            df=df[df["contractType"]=="CO"]
            selected_sp=max(df["strikePrice"].tolist())
            gap = selected_sp-closePrice
            for sp in df["strikePrice"].tolist():
                if sp-closePrice>0 and sp-closePrice<gap:
                    selected_sp= sp
                    gap = selected_sp-closePrice
        else:
            raise ValueError("trading-direction is neither CO or PO, but started trading")

        # 记录换仓日期和新持仓期权合约，以便下文导入Backtest_Engine回测框架（自建的期权回测框架）
        df_selected = df[ df["strikePrice"]==selected_sp]
        secID=df_selected["secID"].values[0]
        ticker=df_selected["ticker"].values[0]
        self.position_table[current_date0]={
            "stock":{},
            "fund":{},
            "options":{ticker:-100*1e4},
            "futures":{}
        }
        
        # 打印每步信息
        string="看多" if direction==1 else "看空"
        print string," 最新义务方期权合约%s," %ticker, "到期日%s," %self.last_tradeDate, "标的物价格%s" %closePrice,"\n"
        
        return df
```

---
【策略逻辑】
1. 上证50指数出现MACD金叉，则看多50ETF净值，具体操作是“卖出浅虚值50ETF近月Put期权”
2. 上证50指数出现MACD死叉，则看空50ETF净值，具体操作是“卖出浅虚值50ETF近月Call期权”
3. 如果当前期权持仓合约即将到期，则移仓换月。这里有个参数lag_days，经过测试在到期日提前3个交易日的开盘时执行移仓换月效果最好。

每一步都打印了上证50的MACD指标是否有金叉死叉，若有则在下一个交易日开盘时刻执行交易操作。大家可自行在通达信、同花顺等软件上自行比对历史数据，是否准确地出现了MACD信号。

```python
# === 数据处理：参数设置部分 ===

start_date = '2016-01-01'     # 回测起始时间
end_date = '2021-10-24'       # 回测结束时间

marketIndex="000016" # 股票指数，用于观察MACD金叉和死叉
varTicker="510050"   # 标的物ETF，用于判断交易哪个执行价的“浅虚值期权”

lag_days=3 # 移仓换月，在最后交易日之前，提前几天执行？


# === 股票指数的MACD数据分析，以及根据（1）MACD金叉死叉 （2）最新持仓是否到了移仓换月的日期 来进行交易信号的生成 ===

# 准备步骤，建立“每个交易日”的“前一个交易日”的索引
df_preTradeDate=DataAPI.TradeCalGet(
    exchangeCD=u"XSHG",beginDate=u"20140101",endDate=u"20291231",isOpen=u"1",
    field=u"calendarDate,prevTradeDate",pandas="1"
).set_index("calendarDate")
#print df_preTradeDate    
 
# 抽取数据，股票指数（此处是 上证50）的日行情和MACD数据，感兴趣者可以与通达信等软件对比，uqer的MACD数据很准确
field="tradeDate,ticker,MACD,Close"
df_macd = DataAPI.MktIdxFactorDateRangeGet(
    beginDate=start_date.replace("-",""),endDate=end_date.replace("-",""),
    ticker=marketIndex,field=field,pandas="1"
)
#print df_macd

# 准备记录，买卖换仓的“全集总信息”，即实例化一个Account类
obj_account=Account(df_preTradeDate)
direction=0

# 主循环开始
for i in range(len(df_macd)): 
    if i<5: continue #初始几个数据点剔除
    
    current_date=df_macd.at[i,"tradeDate"] #当前交易日
    flag_1 = False #记录MACD是否有金叉死叉
    flag_2 = False #记录是否需要移仓换月
     
   
    # 判断：上一个交易日收盘的时候，MACD是否有金叉死叉
    macd_list=df_macd["MACD"].tolist()
    if macd_list[i-1]>0 and macd_list[i-2]<=0:
        flag_1 = True
        direction=1
    elif macd_list[i-1]<0 and macd_list[i-2]>=0:
        flag_1 = True
        direction=-1
    
    # 判断：当前交易日是否已经达到了，移仓换月的日期
    if obj_account.last_tradeDate!=None and current_date>=obj_account.lagDays_dict[lag_days]:
        flag_2 = True
    
    # 打印交易信号的信息，方便大家比对其他软件，辨别计算结果的真伪
    if flag_1 or flag_2:
        if flag_1 and direction==1:
            print current_date, "股票指数MACD金叉"
        elif flag_1 and direction==-1:
            print current_date, "股票指数MACD死叉"
        elif flag_2:
            print current_date, "移仓换月"
        
        # 如果发生了交易信号，就要调用上文Account类中的相关函数，记录交易日期和持仓信息
        df=obj_account.trade_option(current_date,varTicker,direction,lag_days=lag_days)
        #print df

        
# 这个步骤主要是，打印所有持仓信息的“总表格”，在本帖子中暂时被注释掉了，不显示
df0= pd.DataFrame(obj_account.position_table)
df = pd.DataFrame(data =df0.values.T, index=df0.columns, columns=df0.index)
df = df[['options']]
#print df
```

鉴于优矿上一直没有把期权合约加入它的回测框架，导致我只能自己写一个回测框架，结合优矿的历史价格数据进行回测。<em>~ 期权投资策略真的这么非主流吗，伤心 ~</em>

不过，量化投资研究者自行搭建回测框架，也是必然经历的锻炼。本框架来自一个旧帖，详见<a href="https://uqer.datayes.com/v3/community/share/5f26988fc2c58f011dcb11fa"> 《搭建极简版回测框架，用ETF期权对冲股票持仓》</a>

```python
"""这个类用来保存模拟账户里的资金信息和持仓头寸，
是Backtest_Engine的一个属性变量。
如果DIY回测框架时，这个类也是可以扩展些内容的。
"""
class Account_Obj(object):
    def __init__(self,tradeDateList):
        #现金持仓表
        columns=[
            "tradeDate",
            "total_equity",
            "cash",
            "stock",
            "fund",
            "options",
            "futures"
        ]
        self.positionDf=pd.DataFrame(
            data={
            "tradeDate":tradeDateList,
            "total_equity":[None]*len(tradeDateList),
            "cash":[None]*len(tradeDateList),
            "stock":[{}]*len(tradeDateList),
            "fund":[{}]*len(tradeDateList),
            "options":[{}]*len(tradeDateList),
            "futures":[{}]*len(tradeDateList)
            },columns=columns
        ).set_index("tradeDate")
        
        
"""
这个Backtest_Engine类就是回测框架的主体。
比如，天软科技的TSBacktesting也是一个类，大致结构也有类似之处，只是天软做的更好，而这里更简化。
"""
class Backtest_Engine(object):
    ### 类的初始化，生成各种属性
    def __init__(self,positionTable,lastDate,initial_cash=50*1e4*1.0):
        self.beginDate=min(positionTable.keys())
        self.endDate=lastDate
        self.positionTable=positionTable
        
        self.INITIAL_CASH=initial_cash
        self.UNIVERSE={
            "index":["000016","000300"],
            "stock":[],
            "fund":[],
            "options":[],
            "futures":[]
        }
        self.priceTables={
            "index":None,
            "stock":None,
            "fund":None,
            "options":None,
            "futures":None
        }
        self.tradeDate_list=None
        self.get_initial_data()
        
    ### 加载金融数据的函数，这里当然用的是优矿的数据^_^
    def get_initial_data(self):
        # 抽取50指数和300指数，同时这也是交易日序列
        field="tradeDate,openIndex,highestIndex,lowestIndex,closeIndex"
        self.priceTables["index"]=DataAPI.MktIdxdGet(
            ticker=self.UNIVERSE["index"],beginDate=self.beginDate,endDate=self.endDate,
            exchangeCD=u"XSHE,XSHG",field=field
        ).set_index("tradeDate")
        self.tradeDate_list=list(self.priceTables["index"].index.unique())
        self.account=Account_Obj(self.tradeDate_list)
        self.yesterday_position=None
        print "trade-days length:",len(self.tradeDate_list)
        
        # 整理股票池
        self.addInto_universe("stock")
        # 抽取股票池的数据
        if len(self.UNIVERSE["stock"])>0:
            field="tradeDate,ticker,openPrice,highestPrice,lowestPrice,closePrice,secShortName,isOpen"
            df=DataAPI.MktEqudAdjGet(ticker=self.UNIVERSE["stock"],beginDate=self.beginDate,endDate=self.endDate,field=field)
            df=df.sort_values(by=["tradeDate","ticker"]).set_index(["tradeDate","ticker"])
            self.priceTables["stock"]=df
            #print df
        
        # 整理基金池
        self.addInto_universe("fund")
        # 抽取基金池的数据
        if len(self.UNIVERSE["fund"])>0:
            field="tradeDate,ticker,openPrice,highestPrice,lowestPrice,closePrice,secShortName"
            df0=DataAPI.MktFunddAdjBfGet(ticker=self.UNIVERSE["fund"],beginDate=self.beginDate,endDate=self.endDate,field=field)
            df0=df0.sort_values(by=["tradeDate","ticker"]).set_index(["tradeDate","ticker"])
            self.priceTables["fund"]=df0
            #print df0
        
        # 整理期权合约的池子
        self.addInto_universe("options")
        # 抽取期权池的数据
        if len(self.UNIVERSE["options"])>0:
            field="tradeDate,ticker,openPrice,highestPrice,lowestPrice,closePrice,secShortName,openInt"
            df1=DataAPI.MktOptdGet(ticker=self.UNIVERSE["options"],beginDate=self.beginDate,endDate=self.endDate,field=field)
            df1=df1.sort_values(by=["tradeDate","ticker"]).set_index(["tradeDate","ticker"])
            self.priceTables["options"]=df1
            #print df1
        
        # 整理期货合约的池子
        self.addInto_universe("futures")
        # 抽取期货池的数据
        if len(self.UNIVERSE["futures"])>0:
            field="tradeDate,ticker,openPrice,highestPrice,lowestPrice,closePrice,secShortName,openInt"
            df2=DataAPI.MktFutdGet(ticker=self.UNIVERSE["futures"],beginDate=self.beginDate,endDate=self.endDate,field=field)
            df2=df2.sort_values(by=["tradeDate","ticker"]).set_index(["tradeDate","ticker"])
            self.priceTables["futures"]=df2
            #print df2
            
        #== 提示数据初始化完成 ==#   
        print u"历史数据抽取结束"
    
    ### 中间过程函数，用于统计所用到的每一个资产合约
    def addInto_universe(self,equity_label_str):
        if equity_label_str not in ["stock","fund","options","futures"]:
            raise KeyError(u"回测过程出错，在子函数addInto_universe中，资产标签名字有问题")
        for day in sorted(self.positionTable.keys()):
            if day in self.tradeDate_list:
                stock_dict=self.positionTable[day][equity_label_str]
                for stock in sorted(stock_dict.keys()):
                    if stock not in self.UNIVERSE[equity_label_str]:
                        self.UNIVERSE[equity_label_str].append(stock)
        self.UNIVERSE[equity_label_str]=sorted(self.UNIVERSE[equity_label_str])
        
    ###### 回测的计算函数 ###### 
    def run(self,flag_print=True):
        print "开始跑回测\n"
        for tradeDate in self.tradeDate_list:
            # 平移上一个交易日的现金与持仓
            self.yesterday_position=self.account.positionDf.shift(1)
            self.yesterday_position.loc[self.tradeDate_list[0],:]=[self.INITIAL_CASH,self.INITIAL_CASH,{},{},{},{}]
            self.account.positionDf.loc[tradeDate,:]=self.yesterday_position.loc[tradeDate,:]
            
            # 判断本交易日是否有换仓，如果股票、期权、期货在本交易日有换仓则进行处理
            if tradeDate in self.positionTable.keys(): 
                self.conduct_equity_trade(tradeDate,"stock")
                self.conduct_equity_trade(tradeDate,"fund")
                self.conduct_equity_trade(tradeDate,"options")
                self.conduct_equity_trade(tradeDate,"futures")
                
            # 本交易日结尾，结算总资产
            self.calculate_daily_equity(tradeDate)
                
        #==打印结果==#
        if flag_print:
            print self.account.positionDf
        
    ### 中间过程函数，用于对每一笔交易（或称“换仓”）做结算
    def conduct_equity_trade(self,tradeDate,equity_label_str):
        if equity_label_str not in ["stock","fund","options","futures"]:
            raise KeyError(u"回测过程出错，在子函数conduct_equity_trade中，资产标签名字有问题")
        #== 处理资产的换仓 ==#
        today_pos=self.positionTable[tradeDate][equity_label_str]             
        self.account.positionDf.at[tradeDate,equity_label_str]=today_pos
        # 昨日每一只持仓的股票，若被卖出，则现金做结算
        yes_pos=self.yesterday_position[equity_label_str][tradeDate]
        for stock in yes_pos.keys():#昨日有持仓
            if stock not in today_pos.keys():#今日没有持仓
                openP=(self.priceTables[equity_label_str])["openPrice"][(tradeDate,stock)]
                self.account.positionDf.at[tradeDate,"cash"]+=yes_pos[stock]*openP
            elif yes_pos[stock]!=today_pos[stock]:#今日有持仓，也可能会有仓差，即有过买卖交易
                openP=(self.priceTables[equity_label_str])["openPrice"][(tradeDate,stock)]
                diff_pos=today_pos[stock]-yes_pos[stock]
                self.account.positionDf.at[tradeDate,"cash"]-=diff_pos*openP
        for stock in today_pos.keys():#今日有持仓
            if stock not in yes_pos.keys():#昨日没有持仓，即新增持仓
                openP=(self.priceTables[equity_label_str])["openPrice"][(tradeDate,stock)]
                self.account.positionDf.at[tradeDate,"cash"]-=today_pos[stock]*openP

    ### 中间过程函数，用于对每一根K线技术时做相应计算
    def calculate_daily_equity(self,tradeDate): #计算当天总资产   
        total_equity=self.account.positionDf.at[tradeDate,"cash"]
        for equity_label in ["stock","fund","options","futures"]: #股票、基金、期权、期货
            for contract in self.account.positionDf.at[tradeDate,equity_label].keys(): 
                pos=(self.account.positionDf.at[tradeDate,equity_label])[contract]
                closeP=(self.priceTables[equity_label])["closePrice"][(tradeDate,contract)]
                total_equity+=pos*closeP
        self.account.positionDf.at[tradeDate,"total_equity"]=total_equity
    
    ### 简单的画图函数
    def plot(self,figsize=(12,5)):
        ax=self.account.positionDf["total_equity"].plot(figsize=figsize)
        return ax
```

代码让数据运营的搬砖工调试好之后，大家就可以看到“MACD择时指标+滚动逐月卖期权”的策略是否真的有效：
- 基本结论大致符合初始猜想，MACD金叉和死叉信号的择时能力不太可靠。有些月份成功率较高，有些月份会接连亏损
- 总体上看，这个策略是正向收益的策略，即使盈亏比达不到资管产品的要求，但“卖权”的思路值得结合其他方法深化

【结语】<br>

这篇帖子想用一个具体案例告诉大家，利用规范的金融数据平台，整理一个期权投资策略，并不是难事。
- <b>真正难的是什么呢？判断股票指数（或者商品期货价格）的涨跌行情很难，具体来说，是按月度或周度进行多空方向的择时很难。</b><br>

但余力老师尝试过期权策略的数据回测，认为“长期滚动卖出期权”是一种构建高盈亏比的期权策略的合理模式。当然，在大盘走出强趋势行情的月份一定要避免卖权操作。<br><br>
说到底，想做好股票ETF的期权策略，一定要了解上证50、沪深300（和未来可能上期权的中证500、中证1000等）各大指数的涨跌因素。本人从期货圈过渡到股票圈，感到每周影响股票指数的经济事件数量比单一品种的商品期货，要多很多。股票指数的择时，涉及国内海外的宏观经济数据公布、政府的行业政策公布、各大财经媒体关注的重点事件、近期金融圈热议的行业风格轮动等等。
- <b>不过从量化角度来说，“滚动卖出Put期权”策略的成立，只需要假设一条规则：中国50和300指数必然不遵循课本上“对数正态分布”的标准随机运动，而是“上涨概率大于下跌”。它们是最白马蓝筹的一撮股票，似乎我们的国运，理应如此……</b>
