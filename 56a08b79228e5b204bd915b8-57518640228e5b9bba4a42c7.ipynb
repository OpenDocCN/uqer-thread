function:stock_get

输入是股票代码，起始时间，结束时间，所需字段。输出是DataFrame。

比如我要获取平安银行和万科A这两只股票，在20160110-20160120区间内每天的市盈率，证券名称，市净率，最低价，本日融资买入额，那么就像下面这样写：

```python
factor_fields = set(Factor.__dict__)
mkt_fields = ['secShortName', 'exchangeCD','preClosePrice', 'actPreClosePrice',
                     'openPrice', 'highestPrice', 'lowestPrice', 'closePrice', 'turnoverVol', 
                     'turnoverValue', 'dealAmount', 'turnoverRate', 'negMarketValue', 'marketValue', 'isOpen']
fst_fields = ['finVal', 'finBuyVal', 'finRefundVal', 'secVol', 'secSellVol', 'secRefundVol', 'secVal', 'tradeVal']

name_map = {'pre_close': 'preClosePrice',
                       'open': 'openPrice',
                       'high': 'highestPrice',
                       'low': 'lowestPrice',
                       'close': 'closePrice',}

def split_fields(fields):
    return_fields = {'factor':['ticker', 'tradeDate'], 
                              'daily_market':['ticker', 'tradeDate'], 
                              'fst':['ticker', 'tradeDate'],
                              'fst':['ticker', 'tradeDate']}
    for field in fields:
        field = field.replace(' ','')
        field = name_map.get(field, field)
        if field in factor_fields:
            return_fields['factor'].append(field)
        elif field in mkt_fields:
            return_fields['daily_market'].append(field)
        elif field in fst_fields:
            return_fields['fst'].append(field)
    return return_fields

def StockGet(ticker='000001', start_date='20160104', end_date='', field='', adj=True):
    """获取证券的基本信息

        Parameters
        ----------
        ticker : 股票代码，如：'000001'，或：'000001,000002'，['000001', '000002']。
        start_date: 起始日期，如：'20160104'，datetime.datetime(2016,1,4)。
        end_date: 起始日期，如：'20160104'，datetime.datetime(2016,1,4)，默认为当天。
        field: 待查询的证券字段,
        name                               type        desc
        exchangeCD                    str         交易市场
        secShortName                str         证券简称
        secShortNameEn               str         证券英文简称
        tradeDate                          str         交易日期
        preClosePrice/pre_close    float       昨收盘(前复权)
        openPrice/open             float       今开盘(前复权)
        highestPrice/high          float       最高价(前复权)
        lowestPrice/low            float       最低价(前复权)
        closePrice/close           float       今收盘(前复权)
        turnoverVol                float       成交量
        turnoverValue              float       成交金额
        finVal                        float       本日融资余额
        finBuyVal                  float       本日融资买入额
        finRefundVal               float       本日融资偿还额
        secVol                     float       本日融券余量
        secSellVol                 float       本日融券卖出量
        secRefundVol               float       本日融券偿还量
        secVal                     float       本日融券余量金额
        tradeVal                   float       本日融资融券余额
        %s
        adj: 除权标志，True为除权行情，False为非除权行情，默认为True。

        """
    
    if isinstance(field, str):
        field = field.split(',')
        
    if isinstance(ticker, str):
        ticker = ticker.split(',')
        
    fields = split_fields(field)
    dfs = []
    
    factor_fields = fields['factor']
    if factor_fields:
        factor_df = None
        for t in ticker:
            tmp_df = DataAPI.MktStockFactorsDateRangeGet(ticker=t, beginDate=start_date, endDate=end_date, field=factor_fields)
            if factor_df is None: 
                factor_df = tmp_df
            else:
                factor_df = factor_df.append(tmp_df)
            
        dfs.append(factor_df)
    
    fst_fields = fields['fst']
    if fst_fields:
        fst_df = DataAPI.FstDetailGet(ticker=ticker, beginDate=start_date, endDate=end_date, field=fst_fields)
        dfs.append(fst_df)
    
    mkt_fields = fields['daily_market']
    if mkt_fields:
        if adj:
            basic_df = DataAPI.MktEqudAdjGet(beginDate=start_date, endDate=end_date, ticker=ticker, field=mkt_fields)
            dfs.append(basic_df)
        else:
            basic_df = DataAPI.MktEqudGet(beginDate=start_date, endDate=end_date, ticker=ticker, field=mkt_fields)
            dfs.append(basic_df)
 
    return reduce(lambda left,right: pd.merge(left,right,on=['ticker', 'tradeDate']), dfs)

stock_get.__doc__ =  stock_get.__doc__   %('\n'.join([ '%s     float      %s'%(f, Factor.__dict__[f].__doc__) for f in dir(Factor)]) )
```
