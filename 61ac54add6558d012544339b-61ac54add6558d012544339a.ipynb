# pyecharts+期货套保效果分析(F)_Vide

```python
#!/usr/bin/env python
# -*- coding: utf-8 -*-

import pandas as pd
import numpy as np
import scipy
import matplotlib.pyplot as plt
import datetime
from scipy.optimize import leastsq
import math
import random

import pyecharts.options as opts
from pyecharts.charts import Line
from pyecharts.render import make_snapshot
from snapshot_selenium import snapshot


# 模拟其他费用随机游走
def addNewColumns(df,newColumns='A',mean=0,std=0.2,year=252,S0=0.1,aFee=1.25,
                  maxFee=0,minFee=0,maxColumns='',minColumns=''):
    dfColList = [x for x in df.columns]
    df['ri'] = np.random.normal(loc=mean/year, scale=std / (year ** 0.5), size=len(df))
    df['OtherFee'] = [S0 for x in range(len(df))]
    for indexI in range(len(df)):
        if indexI > 0:
            df.loc[indexI, 'OtherFee'] = df.loc[indexI - 1, 'OtherFee'] * math.e ** df.loc[indexI, 'ri']
    df['OtherFee'] = df.loc[:,'OtherFee'].apply(lambda x:x+aFee)
    # print(df)
    # 最大值封顶
    if maxFee!=0:
        if maxColumns!='':
            df[maxColumns] = df.loc[:, 'OtherFee'].apply(lambda x: min(x,maxFee))
            dfColList = dfColList+[maxColumns]
        else:
            df['OtherFee'] = df.loc[:, 'OtherFee'].apply(lambda x: min(x,maxFee))
    # 最小值封顶
    if minFee!=0:
        if minColumns!='':
            df[minColumns] = df.loc[:, 'OtherFee'].apply(lambda x: max(x, minFee))
            dfColList = dfColList+[minColumns]
        else:
            df['OtherFee'] = df.loc[:, 'OtherFee'].apply(lambda x: max(x,minFee))
    df[newColumns] = df.loc[:,'OtherFee']
    df = df.loc[:,dfColList+[newColumns]]
    return df

startDate = '20170101'
startDate = datetime.datetime.strptime(startDate,'%Y%m%d')
endDate = datetime.datetime.now()
dateList = []
IDate = startDate
while IDate<endDate:
    dateList.append(IDate.strftime('%Y%m%d'))
    IDate = IDate+datetime.timedelta(days=1)
else:
    dateList.append(IDate.strftime('%Y%m%d'))

waterFeeList = [10 for x in range(len(dateList))]

# 日期、水费
df = pd.DataFrame([dateList,waterFeeList],index=['Date','Water']).T
print(df)
# 电费
df = addNewColumns(df,newColumns='Electric',mean=0,std=0.02,S0=0.1,aFee=12.5)
# 人工费
df = addNewColumns(df,newColumns='Labor',mean=0.05,std=0.02,S0=7,aFee=30,minFee=35)
# 原材料期货A
df = addNewColumns(df,newColumns='CostFA_nh',mean=0.001,std=0.25,S0=10,aFee=13,
                   maxFee=25,maxColumns='CostFA_h25')
# 原材料期货B
df = addNewColumns(df,newColumns='CostFB_nh',mean=-0.01,std=0.11,S0=16,aFee=20,
                   maxFee=38,maxColumns='CostFB_h38')
# 原材料期货C
df = addNewColumns(df,newColumns='CostFC_nh',mean=0,std=0.4,S0=6,aFee=9,
                   maxFee=20,maxColumns='CostFC_h20')
# 产品期货D
df = addNewColumns(df,newColumns='IncomeFC_nh',mean=0,std=0.3,S0=100,aFee=42,
                   minFee=136,minColumns='IncomeFC_h136')
# print(df.loc[:,['h','nh','IncomeFC_nh','IncomeFC_h136']])

dfColList = [x for x in df.columns]
df2 = df.loc[:,dfColList]
df2['AElectric'] = df2.apply(lambda x:x.Water+x.Electric,axis=1)
df2['ALabor'] = df2.apply(lambda x:x.AElectric+x.Labor,axis=1)
df2['ACostFA_nh'] = df2.apply(lambda x:x.ALabor+x.CostFA_nh,axis=1)
df2['ACostFA_h25'] = df2.apply(lambda x:x.ALabor+x.CostFA_h25,axis=1)
df2['ACostFB_nh'] = df2.apply(lambda x:x.CostFB_nh+x.ACostFA_nh,axis=1)
df2['ACostFB_h38'] = df2.apply(lambda x:x.CostFB_h38+x.ACostFA_h25,axis=1)
df2['ACostFC_nh'] = df2.apply(lambda x:x.ACostFB_nh+x.CostFC_nh,axis=1)
df2['ACostFC_h20'] = df2.apply(lambda x:x.ACostFB_h38+x.CostFC_h20,axis=1)
columnsList = ['Date','Water','AElectric','ALabor','ACostFA_nh',
               'ACostFA_h25','ACostFB_nh','ACostFB_h38','ACostFC_nh','ACostFC_h20']
print(df2.loc[:,columnsList])


def line_base() -> Line:
    c = (
        Line()
            .add_xaxis(df.loc[:, 'Date'].values.tolist())

            .add_yaxis("期货产品D套保",
                       [round(x, 3) for x in df2.loc[:, "IncomeFC_h136"].values.tolist()],
                       is_smooth=True,
                       symbol="triangle",
                       symbol_size=[5, 5],
                       linestyle_opts=opts.LineStyleOpts(color="red", width=1,),
                       markpoint_opts=opts.MarkPointOpts(label_opts=opts.LabelOpts(color="black", position="inside"),
                                                         data=[opts.MarkPointItem(type_="min"),opts.MarkPointItem(type_="max")]),
                       itemstyle_opts=opts.ItemStyleOpts(
                           border_width=1, border_color="blue", color="red"
                       )
                       )
            .add_yaxis("期货产品D非套保",
                       [round(x, 3) for x in df2.loc[:, "IncomeFC_nh"].values.tolist()],
                       is_smooth=True,
                       symbol="rect",
                       symbol_size=[5, 5],
                       linestyle_opts=opts.LineStyleOpts(color="red", width=1,type_="dashed"),
                       markpoint_opts=opts.MarkPointOpts(label_opts=opts.LabelOpts(color="black", position="inside"),
                                                         data=[opts.MarkPointItem(type_="min"),opts.MarkPointItem(type_="max")]),
                       itemstyle_opts=opts.ItemStyleOpts(
                           border_width=1, border_color="blue", color="red"
                       )
                       )

            .add_yaxis("期货原料C非套保",
                       [round(x, 3) for x in df2.loc[:, "ACostFC_nh"].values.tolist()],
                       is_smooth=True,
                       symbol="triangle",
                       symbol_size=[5, 5],
                       linestyle_opts=opts.LineStyleOpts(color="blue", width=1, type_="dashed" ),
                       markpoint_opts=opts.MarkPointOpts(label_opts=opts.LabelOpts(color="black", position="inside"),
                                                         data=[opts.MarkPointItem(type_="min"),
                                                               opts.MarkPointItem(type_="max")]),
                       itemstyle_opts=opts.ItemStyleOpts(
                           border_width=1, border_color="blue", color="blue"
                       )
                       )
            .add_yaxis("期货原料C套保",
                       [round(x, 3) for x in df2.loc[:, "ACostFC_h20"].values.tolist()],
                       is_smooth=True,
                       symbol="rect",
                       symbol_size=[5, 5],
                       linestyle_opts=opts.LineStyleOpts(color="blue", width=1),
                       markpoint_opts=opts.MarkPointOpts(label_opts=opts.LabelOpts(color="black", position="inside"),
                                                         data=[opts.MarkPointItem(type_="min"),
                                                               opts.MarkPointItem(type_="max")]),
                       itemstyle_opts=opts.ItemStyleOpts(
                           border_width=1, border_color="blue", color="blue"
                       )
                       )

            .add_yaxis("期货原料B非套保",
                       [round(x, 3) for x in df2.loc[:, "ACostFB_nh"].values.tolist()],
                       is_smooth=True,
                       symbol="triangle",
                       symbol_size=[5, 5],
                       linestyle_opts=opts.LineStyleOpts(color="yellow", width=1, type_="dashed" ),
                       markpoint_opts=opts.MarkPointOpts(label_opts=opts.LabelOpts(color="black", position="inside"),
                                                         data=[opts.MarkPointItem(type_="min"),
                                                               opts.MarkPointItem(type_="max")]),
                       itemstyle_opts=opts.ItemStyleOpts(
                           border_width=1, border_color="blue", color="yellow"
                       )
                       )
            .add_yaxis("期货原料B套保",
                       [round(x, 3) for x in df2.loc[:, "ACostFB_h38"].values.tolist()],
                       is_smooth=True,
                       symbol="rect",
                       symbol_size=[5, 5],
                       linestyle_opts=opts.LineStyleOpts(color="yellow", width=1),
                       markpoint_opts=opts.MarkPointOpts(label_opts=opts.LabelOpts(color="black", position="inside"),
                                                         data=[opts.MarkPointItem(type_="min"),
                                                               opts.MarkPointItem(type_="max")]),
                       itemstyle_opts=opts.ItemStyleOpts(
                           border_width=1, border_color="blue", color="yellow"
                       )
                       )

            .add_yaxis("期货原料A非套保",
                       [round(x, 3) for x in df2.loc[:, "ACostFA_nh"].values.tolist()],
                       is_smooth=True,
                       symbol="triangle",
                       symbol_size=[5, 5],
                       linestyle_opts=opts.LineStyleOpts(color="purple", width=1, type_="dashed" ),
                       markpoint_opts=opts.MarkPointOpts(label_opts=opts.LabelOpts(color="black", position="inside"),
                                                         data=[opts.MarkPointItem(type_="min"),
                                                               opts.MarkPointItem(type_="max")]),
                       itemstyle_opts=opts.ItemStyleOpts(
                           border_width=1, border_color="blue", color="purple"
                       )
                       )
            .add_yaxis("期货原料A套保",
                       [round(x, 3) for x in df2.loc[:, "ACostFA_h25"].values.tolist()],
                       is_smooth=True,
                       symbol="rect",
                       symbol_size=[5, 5],
                       linestyle_opts=opts.LineStyleOpts(color="purple", width=1),
                       markpoint_opts=opts.MarkPointOpts(label_opts=opts.LabelOpts(color="black", position="inside"),
                                                         data=[opts.MarkPointItem(type_="min"),
                                                               opts.MarkPointItem(type_="max")]),
                       itemstyle_opts=opts.ItemStyleOpts(
                           border_width=1, border_color="blue", color="purple"
                       )
                       )
            .add_yaxis("人工",
                       [round(x, 3) for x in df2.loc[:, "ALabor"].values.tolist()],
                       is_smooth=True,
                       symbol="rect",
                       symbol_size=[5, 5],
                       linestyle_opts=opts.LineStyleOpts(color="green", width=1),
                       markpoint_opts=opts.MarkPointOpts(label_opts=opts.LabelOpts(color="black", position="inside"),
                                                         data=[opts.MarkPointItem(type_="min"),
                                                               opts.MarkPointItem(type_="max")]),
                       itemstyle_opts=opts.ItemStyleOpts(
                           border_width=1, border_color="blue", color="green"
                       )
                       )
            .add_yaxis("电费",
                       [round(x, 3) for x in df2.loc[:, "AElectric"].values.tolist()],
                       is_smooth=True,
                       symbol="rect",
                       symbol_size=[5, 5],
                       linestyle_opts=opts.LineStyleOpts(color="orange", width=1),
                       markpoint_opts=opts.MarkPointOpts(label_opts=opts.LabelOpts(color="black", position="inside"),
                                                         data=[opts.MarkPointItem(type_="min"),
                                                               opts.MarkPointItem(type_="max")]),
                       itemstyle_opts=opts.ItemStyleOpts(
                           border_width=1, border_color="blue", color="orange"
                       )
                       )
            .add_yaxis("水费",
                       [round(x, 3) for x in df2.loc[:, "Water"].values.tolist()],
                       is_smooth=True,
                       symbol="rect",
                       symbol_size=[5, 5],
                       linestyle_opts=opts.LineStyleOpts(color="blue", width=1),
                       markpoint_opts=opts.MarkPointOpts(label_opts=opts.LabelOpts(color="black", position="inside"),
                                                         data=[opts.MarkPointItem(type_="min"),
                                                               opts.MarkPointItem(type_="max")]),
                       itemstyle_opts=opts.ItemStyleOpts(
                           border_width=1, border_color="blue", color="blue"
                       )
                       )
            .set_series_opts(
            areastyle_opts=opts.AreaStyleOpts(opacity=0.5),
            label_opts=opts.LabelOpts(is_show=False),
        )
            .set_global_opts(
            title_opts=opts.TitleOpts(
                title="当日期货产品成本利润构成",
                pos_bottom="1%",
                pos_left="center",
                title_textstyle_opts=opts.TextStyleOpts(color="black", font_size=16),
            ),
            datazoom_opts=[opts.DataZoomOpts(), opts.DataZoomOpts(type_="inside"), opts.DataZoomOpts(orient="vertical")],
            toolbox_opts = opts.ToolboxOpts()
        )
    )
    return c

line_base().render("D:\\pyecharts_fakeFuturn_" + startDate.strftime('%Y%m%d') + "_" + endDate.strftime('%Y%m%d') + "_.html")

# make_snapshot(snapshot, line_base().render(), path + "当日蛋内份额构成.png")
```
