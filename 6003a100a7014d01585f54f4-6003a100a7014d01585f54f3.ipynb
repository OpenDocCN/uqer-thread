# 上证指数以及无风险收益率底线

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
SH_index = DataAPI.MktIdxdGet(indexID=u"",ticker=u"000001",tradeDate=u"",beginDate=u"19901220",
                              endDate=u"",exchangeCD=u"XSHE,XSHG",field=u"",pandas="1")
SH_index.head()
```

```python
#从开始以来的上证指数
SH_price = SH_index[['tradeDate','preCloseIndex']]
SH_price.loc[-1] = ['1990-12-19',100]  # adding a row
SH_price.index = SH_price.index + 1  # shifting index
SH_price = SH_price.sort_index()
```

```python
#判断月初
SH_price['ismonth_beg'] = 0
for i in SH_price.index[1::]:
    if float(SH_price.iloc[i,0][8:10]) < float(SH_price.iloc[i-1,0][8:10]):
        SH_price.iloc[i,2] = 1
    else:
        SH_price.iloc[i,2] = 0
SH_price.iloc[0,2] = 1
SH_price['1000块钱'] = SH_price['ismonth_beg'].apply(lambda x: 1000 if x==1 else 0)
```

```python
SH_price['riskfree_rate'] = 1.00017
SH_price['无风险理财净值'] = SH_price['riskfree_rate'].cumprod()
SH_price['无风险理财单次购买单位'] = SH_price['1000块钱']/SH_price['无风险理财净值']
SH_price['无风险理财单位'] = SH_price['无风险理财单次购买单位'].cumsum()
SH_price['无风险理财价值'] = SH_price['无风险理财单位']*SH_price['无风险理财净值']
SH_price['指数基金净值'] = SH_price['preCloseIndex']/1000
SH_price['指数基金单次购买单位'] = SH_price['1000块钱']/SH_price['指数基金净值']
SH_price['指数基金单位'] = SH_price['指数基金单次购买单位'].cumsum()
SH_price['指数基金价值'] = SH_price['指数基金单位']*SH_price['指数基金净值']
SH_price.head()
```

```python
threeline = SH_price[['tradeDate','无风险理财价值','指数基金价值','preCloseIndex']]
fig = plt.figure(figsize=(18,10))
ax = fig.add_subplot(111)
ax.plot(threeline['指数基金价值'], '-', label = 'stock index fund',color='white')
ax.plot(threeline['无风险理财价值'], '-', label = 'risk free fund')
ax2 = ax.twinx()
ax2.plot(threeline['preCloseIndex'], '-r', label = 'stock index')
ax.legend(loc=2)
ax.grid()
ax.set_xlabel("Time")
ax.set_ylabel(r"fund value")
ax2.set_ylabel(r"stock index")
#ax2.set_ylim(0, 6000)
#ax.set_ylim(0,150000)
ax2.legend(loc=0)
```
