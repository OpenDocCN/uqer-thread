# RSI指标与未来N个交易日累计涨跌幅的相关系数分析

简单验证下RSI指标作为超卖超买指标的前瞻性

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import talib as ta
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from mpl_toolkits.mplot3d import Axes3D
```

```python
start_date='19910101'
end_date='20160415'
fields=['tradeDate','preCloseIndex','openIndex','lowestIndex','highestIndex','closeIndex','turnoverVol','turnoverValue','CHGPct']
index_data=DataAPI.MktIdxdGet(ticker='000001',beginDate=start_date,endDate=end_date,field=fields)
```

```python
def get_corr_rsi_pct(rsi_period,pct_period):
    #若用T日数据计算rsi，则应从T+1日开始计算未来N天的涨跌幅，从而避免使用未来数据
    index_data['pct_'+str(pct_period)]=(index_data['closeIndex'].shift(pct_period)-index_data['openIndex'].shift(1))/index_data['closeIndex']
    index_data['rsi']=ta.RSI(np.array(index_data['closeIndex']),timeperiod=rsi_period)
    return index_data['pct_'+str(pct_period)].corr(index_data['rsi'])
#画出rsi与pct的相关系数曲面
rsi_timeperiod=np.arange(5,25)
pct_shiftperiod=np.arange(5,30)
corr_array=np.empty([len(pct_shiftperiod),len(rsi_timeperiod)])
rsi_period,pct_period=np.meshgrid(rsi_timeperiod,pct_shiftperiod)
for i in range(len(pct_shiftperiod)):
    for j in range(len(rsi_timeperiod)):
        corr_array[i,j]=get_corr_rsi_pct(rsi_period[i,j],pct_period[i,j])
```

```python
#作图                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                 
fig = plt.figure(figsize=(15, 12))
ax = fig.gca(projection='3d')
surf = ax.plot_surface(rsi_period, pct_period, corr_array, rstride=2, cstride=2,cmap=plt.cm.coolwarm, linewidth=0.5,antialiased=True)
ax.set_xlabel('rsi_period')
ax.set_ylabel('pct_period')
ax.set_zlabel('corr')
ax.set_title('corr of rsi and pct_change')
fig.colorbar(surf, shrink=0.5, aspect=5)
```

```python
#RSI选取的窗口期长，对预测未来较长一段时间的大盘走势较有帮助
corr_pd=pd.DataFrame(corr_array,columns=rsi_timeperiod)
corr_pd.set_index(pct_shiftperiod)
ax1=corr_pd.plot(figsize=(16,12))
ax1.set_xlabel("pct_shiftperiod")
ax1.set_ylabel('corr')
ax1.legend(loc=3)
```
