广播
两个数组的维度不同，numpy可以通过广播的方式实现对形状不相似的数组进行操作。较小的数组会广播到最大的数组中。

```python
a = np.array([1,2,3])    #1*3的矩阵
b = np.array([4,5])     #1*2的矩阵
#两者无法相乘
print np.reshape(a,(3,1))*b
#将其变成3*1的矩阵，就可以跟b相乘了。
```

```python
#将向量b加到矩阵a行上

a = np.arange(12).reshape(4,3)
b = np.array([1,0,1])
print a
print b
print a + b
```

```python
#将向量b加到矩阵a列上

a = np.arange(12).reshape(3,4)
b = np.array([1,0,1])
#转制在转制
print (a.T + b).T
```

```python
print a
print a*5    #四则运算都可以
```

scipy的几个模块具有多种功能。
概率统计方法：stats模块

```python
#以正态分布为例
from scipy.stats import norm
import numpy as np
from matplotlib import pylab   #为了画面展示
```

```python
#随机产生5000个点，满足正态分布
x_norm = norm.rvs(size = 5000, loc = 0.5, scale = 0.5)
```

```python
#在这组数据下，用极大似然估计正态分布的参数
x_mean, x_std = norm.fit(x_norm)
print 'mean: %s, std: %s'%(x_mean, x_std)
```

优化模块：optimize

```python
from scipy import optimize

def f(x):
    return x**2 +10 *x +10
```

```python
x = np.arange(-10,0, 0.1)
pylab.plot(x,f(x))
```

```python
#求解
print optimize.fsolve(f, 0 )   #寻找最近的根
print optimize.fsolve(f, -7 )
```

```python
#求最小值
optimize.fmin_bfgs(f,0)
```

```python
optimize.minimize(f, 0).x
```

```python
#设置约束条件
#使其定义域大于等于-4
cons = ({'type': 'ineq','fun':lambda x: x+4})
```

```python
optimize.minimize(f,0,constraints = cons).x
```
