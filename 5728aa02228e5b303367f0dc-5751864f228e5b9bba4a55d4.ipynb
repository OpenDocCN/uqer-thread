Obtaining historical stock and index data

```python
# request the previous three years of data for 600630.XSHG,000565.XSHE,002736.XSHE,300315.XSHE
# 2012 through 2015
# start end end dates
start = '2012-08-01'
end   = '2015-08-01'

# load the data

ltgf = DataAPI.MktEqudGet(secID='600630.XSHG',beginDate=start,endDate=end,field=u"tradeDate,openPrice,highestPrice,lowestPrice,closePrice,turnoverVol",pandas="1")
ltgf['tradeDate'] = ltgf.apply(lambda row: (datetime.datetime.strptime(row["tradeDate"], "%Y-%m-%d")), axis = 1) #因为DataAPI所有日期都是以字符串形式存储的，因此此处需要转换成日期时间型,这里很重要！
ltgf = ltgf.set_index('tradeDate')#转换后index是DatetimeIndex型

ltgf[:5]
```

```python
# retrieves the historical prices for a list of stocks
def get(tickers, start, end):
    def data(ticker):
        # load the data
        ticker_data = DataAPI.MktEqudGet(secID=ticker,beginDate=start,endDate=end,field=u"tradeDate,openPrice,highestPrice,lowestPrice,closePrice,turnoverVol",pandas="1")
        ticker_data['tradeDate'] = ticker_data.apply(lambda row: (datetime.datetime.strptime(row["tradeDate"], "%Y-%m-%d")), axis = 1) #因为DataAPI所有日期都是以字符串形式存储的，因此此处需要转换成日期时间型,这里很重要！
        ticker_data = ticker_data.set_index('tradeDate')#转换后index是DatetimeIndex型
        return ticker_data        
    datas = map(data, tickers)    
    return pd.concat(datas, keys=tickers, 
                     names=['secID', 'tradeDate'])
```

```python
# get data for the following stocks

universe = ['600630.XSHG','000565.XSHE','002736.XSHE','300315.XSHE']
all_data = get(universe, start, end)
all_data[:5]
```

Fundamental Financial Calculations

```python
# calc % change from day 0 to day 1
AA_p_t0 = daily_close_px.iloc[0]['600630.XSHG']  #Pt-1
AA_p_t1 = daily_close_px.iloc[1]['600630.XSHG']  #Pt
r_t1 = AA_p_t1 / AA_p_t0 - 1 # returns
AA_p_t0, AA_p_t1, r_t1 
```

```python
# we can apply this to everything with the following using slices
dpc_1 = daily_close_px.iloc[1:] / \
        daily_close_px.iloc[:-1].values - 1
dpc_1.ix[:,'600630.XSHG':'300315.XSHE']
```

```python
# show the DataFrame that is the numerator
price_matrix_minus_day1 = daily_close_px.iloc[1:]
price_matrix_minus_day1[:5]
```

```python
# the numerator is a 2-d array, but excludes the last day
daily_close_px.iloc[:-1].values
```

```python
# or using the shift function
dpc_2 = daily_close_px / daily_close_px.shift(1) - 1
dpc_2.ix[:,0:2][:5]
```

```python
# to make this easy, pandas has .pct_change() baked in
daily_pct_change = daily_close_px.pct_change()
daily_pct_change.ix[:,0:2][:5]
```

```python
# set NaN's to 0 
daily_pct_change.fillna(0, inplace=True)
daily_pct_change.ix[:5,:5]
```

Calculating simple daily cumulative returns

```python
# calc the cumulative daily returns 
cum_daily_return = (1 + daily_pct_change).cumprod()
cum_daily_return.ix[:,:2][:5]
```

Comparison of daily percentage change between stocks

```python
def render_scatter_plot(data, x_stock_name, 
                        y_stock_name, xlim=None, ylim=None):
    fig = plt.figure(figsize=(12,8))
    ax = fig.add_subplot(111)
    ax.scatter(data[x_stock_name], data[y_stock_name])
    if xlim is not None: ax.set_xlim(xlim)
    ax.autoscale(False)
    # horiz and v lines at 0
    ax.vlines(0, -10, 10)
    ax.hlines(0, -10, 10)
    # this line would be perfect correlation
    ax.plot((-10, 10), (-10, 10))
    # label axes
    ax.set_xlabel(x_stock_name)
    ax.set_ylabel(y_stock_name)
```

```python
# '600630.XSHG' vs '300315.XSHE'
limits = [-0.15, 0.15]
render_scatter_plot(daily_pct_change, '600630.XSHG', '300315.XSHE', xlim=limits)
plt.savefig('S_04_14.png', bbox_inches='tight', dpi=300)
```
