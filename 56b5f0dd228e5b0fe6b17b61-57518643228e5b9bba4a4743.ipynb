# 沪深300等权重策略

策略介绍：最简单的沪深300等权重策略，每20个交易日做一次rebalance，若可交易的股票数量为N，N小于300，那么每个股票的权重为1/N。看上去效果比我想的要好很多，估计是因为300里面有很多权重股，权重股的股息率较高，所以股息再投资的影响较大。
