```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.optimize as sco
#随便取了三支股票
data1=DataAPI.MktEqudGet(secID=u"",ticker=u"002043",beginDate=u"20130501",endDate=u"20150701",isOpen="",field=u"secID,secShortName,tradeDate,openPrice,highestPrice,lowestPrice,closePrice,turnoverVol",pandas="1")
data2=DataAPI.MktEqudGet(secID=u"",ticker=u"600050",beginDate=u"20130501",endDate=u"20150701",isOpen="",field=u"secID,secShortName,tradeDate,openPrice,highestPrice,lowestPrice,closePrice,turnoverVol",pandas="1")
data3=DataAPI.MktEqudGet(secID=u"",ticker=u"600019",beginDate=u"20130501",endDate=u"20150701",isOpen="",field=u"secID,secShortName,tradeDate,openPrice,highestPrice,lowestPrice,closePrice,turnoverVol",pandas="1")#换了一段时间，10年到12年的时候，三支股票年收益率都为负。。接下来就不好做了。。选股和选时间
data1=data1['closePrice']
ret1=np.log(data1/data1.shift(1))

data2=data2['closePrice']
ret2=np.log(data2/data2.shift(1))

data3=data3['closePrice']
ret3=np.log(data3/data3.shift(1))

dt=pd.DataFrame({'002043':ret1,'600050':ret2,'600019':ret3})
#计算了均值和协方差矩阵
dt.mean()*252
dt.cov()*252
rets=[]
vols=[]
#以2500个点画了了一个不同股票分配比例下的预期收益的波动率，还有夏普（这里无风险收益按0来简化了）
for i in range(2500):
    weights=np.random.random(3)
    weights/=np.sum(weights)
    rets.append(np.sum(dt.mean()*252*weights))
    vols.append(np.sqrt(np.dot(weights.T,np.dot(dt.cov()*252,weights))))
rets=np.array(rets)
vols=np.array(vols)
plt.scatter(vols,rets,c=rets/vols)#错误unsupported operand type(s) for /: 'list' and 'list',得先将vols,rets变成数组
#scatter中第三个数据集是可以通过颜色来可视化的！！！死国一
plt.xlabel('expected volatility')
plt.ylabel('expected return')
plt.colorbar(label='sharpe ratio')
#接下来就是要求最优的选择了，我是选择使夏普最大，也就是负的夏普最小
def statistics(weights):
    weights=np.array(weights)
    rets=np.sum(dt.mean()*252*weights)
    vols=np.sqrt(np.dot(weights.T,np.dot(dt.cov()*252,weights)))
    return np.array([rets,vols,rets/vols])
def min_sharpe(weights):
    return -statistics(weights)[2]
cons=({'type':'eq','fun':lambda x:np.sum(x)-1})#约束条件，各比例加和为1
bnds=({(0,1),(0,1),(0,1)})#各个比例的上下限，0~1
a=[0.4,0.4,0.2]#最初的比例，必须要有就随便写了
opts=sco.minimize(min_sharpe,a,method='SLSQP',bounds=bnds,constraints=cons)#问题是我这个最优化出了问题，虽然是按书上抄来的，但是我连问题是什么都看不懂（捂脸）
opts
```
