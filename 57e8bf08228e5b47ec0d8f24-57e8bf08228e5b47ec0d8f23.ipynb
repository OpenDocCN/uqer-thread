# 关于多期投资、长期超额收益的最近思考
-----
最近一直在思考低风险低收益投资、和高风险高收益投资的权衡问题。
众所周知，风险和收益一直都是两个欢喜冤家，我们只能进行trade off。

另外，我们一直以来思考的框架，都是在"单期投资"的框架下，而投资从来就不是单一时期，而是始终变化着的。
从此也演化出了近年来比较热门的"多期投资"框架，也即多时期下的最优投资组合问题。有意思的是，多期最优投资策略往往和单期最优投资策略不一致。

这方面我们中国人有很多贡献。比较有名的有香港中文大学SEEM系的李端教授，他把CAPM模型在多时期做了扩展，并经过复杂计算得到了解析解。


-------
这里先不讨论李教授的模型。我想起 《investment science 第二版-卢恩伯格》  一书时候，里面也有提到多期投资问题。里面提到一个叫做`对数效用模型`。这里我简单介绍一下。它从一个很有意思的游戏开始。假设有一个轮盘，轮盘上分成3个区域，每个区域面积不同，，且对应不同奖励。你可以在任何区域下注，下注金额是1块。如果最后指针指向不同区域则无法得到奖励且1块钱无法收回。 

第一个区域面积是1/2, 获得的奖励是3块，第二个区域面积是1/3,获得的奖励是2块，第三个区域面积是1/6,获得的奖励是6块。

如果是你来玩这个游戏，初始有100块钱，怎么玩才能在N期之后的期望最大收益？

这时候我们心理就开始算了：第一个的期望收益是1.5 -1 = 0.5，第二个的期望收益是0.667 - 1 = - 0.33，第三个的期望收益是 1 - 1 = 0。那还不简单，就专注于投资第一个区域呗！

这时候，有的同学可能还会考虑，用凯利策略来为每次投资第一个区域的资金进行最优分配。毕竟凯利策略也是多期投资下的嘛。但不管怎么样，你都会想去只投资第一个对不对？


来看看对数效用模型是怎么做的。该模型是把投资者的效用函数假设为对数形式，在此假设下，目标是最大化第N期的期望效用。

这个模型的结论惊人：你只需要最大化第一个时期的效用函数，并保持这个投资比例，就能最大化第N期的期望效用。而且，甚至最后你都不需要效用函数是对数形式的假设！

也就是说，你只需要像对待单期投资一样，对待多期投资。

好吧，是不是感觉好像说了和没说一样，并没有什么特别的感觉？

那么，我们再来看一下前面提到的那个游戏。在这个理论下，给出的策略很违反我们的直觉：始终把1/2的资金投资在面积为1/2的区域，把1/3的资金投资在面积为1/3的区域，1/6的资金投资在面积为1/6的区域。即使是在预期净收益为负数的第二个区域，也进行了投资。

不相信的同学，可以用excel或是其他软件简单蒙特卡洛模拟一下50期、100期、200期之后的结果，对比你的策略和这个对数策略，相信你会很吃惊。

之所以这样投资能长期赚钱的缘由，不完全是因为投资风险分散化，而是由于书中所称的”泵吸"效应，通俗来说就是高抛低吸。哈哈是不是像某强国的庄家所为。。。。

`结论`：
- 如果投资的期限短，那么就投低风险低收益的；如果投资期限长，就投高风险高收益的。这样有较大概率保证，在投资期限末尾的期望价值最大。

文中给的例子显示，在90%的置信区间中，如果期限较长，那么高风险高收益的投资品，必然会比低风险低收益的投资品，有更高的期望回报。

大家可能又会觉得，这不是废话么？有啥意义？


- 我觉得，意义就是，当我们测试各种选股策略，比如多因子策略，我们最后总是能跑出远远高于基准大盘的收益率。包括Fama三因子模型，也让我们选择小市值、高账面市值比的股票。这其实，不就是因为他们都是选择的高风险、高收益的股票种类吗？而且恰好回测时期比较长，那么，跑赢大盘就不奇怪了。

alpha策略，说到底，就是如何找到更高风险、更高收益的股票。这里突出一个"更"字。找到了，那么你跑10年，必然是高于基准远远的。

所以，为什么神创板上诞生了一些收益率高得吓人的明星基金经理，这就不奇怪了。

因此，因子选股，可以分成3类：目标是短期的，选择低风险因子。目标是中期，选择中风险因子。目标是长期，选择高风险因子。


- 当你找到了处于有效前沿上的高风险高收益资产，而且你恰好想投资较长时间，那么就持有吧！！看起来是赌博，实际上并不一定。虽然回撤风险有时很吓人，但如果你本金已经从1万涨到100万，你还担心50%的回撤吗？
