# 我交易风格的形成

    这是我在优矿的第九篇文章，绝对干货。
    我1981年出生于河南省郑州市上街区，一个国有特大型企业里，爷爷是这个企业的总经理，父亲是这个企业下属一个分厂的厂长（我长大了才当的），从小我就过着无忧无虑的生活，我自诩为贵族。
    在我的印象里，爷爷是个十分严谨的老人，像个学者，科研，管理，毛笔字，钓鱼就是他生活的全部，也经常带着我出去玩，爸爸则是一个标准的太子爷，琴棋书画样样精通，聪明，有能力，有自己独立的思维，但从不媚俗，不会巴结别人，所以是个领导，但总是得不到重用。爷爷95年去世了，爸爸觉得没什么好混的了，所以98年就自动下岗了，去了姑姑们所在的城市-深圳。
    我2002年郑州大学化学系化工专业毕业，妈妈也从职工大学校长的职位退了下来，追随爸爸去了深圳，这里多说一句，妈妈是一个工人的女儿，但是学习成绩特别好，为人处世也很好，一直以来都是最优秀的。
    我大学的时候，妈妈就开始炒股了，一开始是让懂得人代买，因为只有我和妈妈在家，所以每次赚钱了都回来跟我分享，亏了钱也回来跟我分析，但是在我印象里那几年股市好像不错，赚了些钱。大四的时候，由于毕业设计不忙，我就开始帮妈妈看股票，加上我爸爸在深圳，那时刚流行银券通，他们就开了账户，但是电脑操作水平一般，我就开始帮他们买卖股票了，我家有电脑很早了，上网也很早，是国内第一批有自己电脑和上网的人，我有个5位数的QQ：65060（微信同号）现在还在使用。
    2002年我考上了南京工业大学的生物化学专业研究生，师从欧阳老师，1年后，由于我自己的原因（考试作弊），被开除了学籍，于是我就开始找工作，由于喜欢证券，加上在深圳开了银券通这个新业务，所以第一份工作是被我的股友，南京国信证券的营销部经理拉着，进入南京国信证券，担任营销部主管职位，负责销售团队招聘，产品宣传策划，员工培训，银行开发等工作，我们从无到有，建立了公司全新营销体系，招聘了一个50人左右的团队，开发了南京所有的银行，服务了很多的客户，在此期间，我学习了各种营销和团队管理手段，积累了一批客户，并有幸认识了一些目前国内有名的庄家和职业操盘手，还有一大批愿意投入证券事业的大学生（目前都是南京证券行业的顶尖人才），当然自己也在股票的世界里开始折腾，这个时候，我的交易风格主要是跟大户。期间也做了一些外汇倒手，赚了一小笔。
    1年半以后，我跟着我的老板转战到南京华泰证券，担任市场部经理职位，又干了1年，在这里，我和几个职业的分析师成为了铁杆朋友，以致于现在我在电视里看到分析师就想笑，因为分析师=骗子，我从不相信他们的分析，随着职位的升高，开始跟着老板有一些上市公司调研的活动，我的交易从跟大户，慢慢的加入了一些自己的价值分析和判断。
    其实我不太适合营销或市场方面的工作，加上股票上小有斩获，有积累了几个大客户，认识了几个好朋友，因此我就脱离了证券公司，和几个朋友成立了一家投资公司，主要业务有两个，一是为上市公司寻找资本，配合一级增发和二级市场股票操作，二是代客理财，由于全流通的到来，这两个业务让我赚了不少的钱，在南京买了属于自己的房子，这个时期，国内出现了一种新的金融衍生品，权证，我发掘了其中巨大的机会，成为了一个完全意义上的量化投资者。
    研究生肄业后，因为我小的时候就很怕我父亲，我一直欺骗我父亲还在上学，毕业后留在学校当辅导员，但其实我2003年就参加了工作，2007年，我研究生肄业和开公司这个消息终于传到我父亲耳朵里，他认为我做的事业不靠谱，我也经过了挣扎，和他大吵，几个月不联系等手段，气的我父亲得了癌症，于是我变卖了南京的公司股份，卖掉了南京的房子，回到了深圳。
    半年后，父亲好了，我随了父亲的心愿，被安排到姑姑的环保公司当了一名环保工程师，我知道家人的良苦用心，但是做过金融再去做其他确实没什么意思，但我还是认真的干了4年，其中我结婚了，娶了一个一个人在深圳打拼的湖南妹子，数学系毕业的聪明的女孩，好多人都说我妈妈是她妈妈，我是女婿，我在想我是不是就是按照妈妈的标准找的那，我很爱她，也很珍惜我们拥有的一切。这其中，我还在炒股，权证没了，股票有赚有亏。
    2012年，我在她的支持下辞了职，继续以股票为生，我发现我跟金融圈有些脱节了，我再也没有以前的意气风发，4年的工程师生涯没给我带来什么，只是随了家人的心愿（现在我不这么想了，其实我当时继续干下去，现在肯定是这个公司的总经理和这个领域的专家，不过现在其实我想回去，还是随时回去，估计1年后也能当上总经理，因为姑姑已经60多了，唯一的儿子在国外）。我只能从以前的笔记中和来深圳开会的原来的同事口中记忆其当年的辉煌，半死不活，我继续价值分析，但是赚钱不多，小市值概念股涨的最快。
    2013年，我觉得只靠股票赚钱不够，因此就又成立了一个网络公司，有幸认识了一个香港人，计算机技术牛人，我们在个人领域从微信公众平台的开发着手，企业领域从环保企业erp进军，虽然没赚什么大钱，但基本够养家糊口，积累了一些研发经验，但是我发现我越来越不喜欢营销了，很多时候都是采购方催着我签合同。这期间，股票基本没什么进展，我开始进军期货行业，被一个我认为的高手骗子骗了10万块，到现在为止，我也不知道到底是怎么没有的。
    2003年底，我太太怀孕了，我也发现我的缺点，不喜欢销售，因此将企业交给合伙人，筹集了200万，打算转战到一个全新的领域，对冲投资，当时设计是国内股票和香港同名股票，股指期货和股票组合进行对冲，设计的思路没有问题，但是执行起来，盈利较慢，因为略有盈利，但心高气傲，将钱全部投资于商品期货领域，由于小孩出生较忙，亏得是血本无归，这期间，因为接触到期货，所以和香港合伙人还在量化领域折腾了一番，积累了量化投资的大量经验，在对冲方面基本没什么问题，在高频方向也小有斩获。
    2015年初，我是一分钱都没有了，气的太太差点和我离婚，不过她也知道我的能力和运气，加上女儿小，所以还是选择继续支持我，我们卖掉了结婚时候我父母给她买的婚房（现在住在政府的公租房内），又筹集了50万，换成我当狗头军师，我太太操作（本来我太太就是学数学的，从小跟他父亲看股票有兴趣，加上这么多年，我的潜移默化，加上她现在在学心理咨询，她已经厉害过我）2015年上半年运气好，赚了些钱，但是下半年，别提了，没赚钱就是亏钱，因为我之前的200万有一部分不是自己的，要还给别人，每月大约有5万的支出，加上养家糊口的2万，所以10月的时候只剩下30万了，我慌神了，因为这个市场30万是不够每月赚6-8万的。
    我得了很严重的抑郁症，每天2点睡觉，4点起床，总是想念以前的婚房（因为公租房比较偏僻，很后悔）我不敢告诉其他人，不愿意家人为我分担，因为其实从我辞职那天起，我父亲就基本不理我了，妈妈虽然很爱我，但是也没有办法，老婆已经在年初跟我闹过一次了，到剩下30万的时候，她也没话说了，女儿1岁半，我很想好好陪伴她，但是我有心无力，身边的朋友不再联系，见人就躲，不吃饭，不睡觉，我只有一个信念，为了家人，我不能死，我不能让他们替我抗应该我承担的一切，虽然我也没有什么办法，只能先赖着，除此之外，生不如死，期间因为经验，我太太成功躲过了1月的股灾，而且这几个月居然没亏，另外其实她说看出我得病，但是没点破。我在做期货的时候就一直很关注期权，但是2015年2月期权开放的时候，我已经没钱了，但是因为兴趣还是开了个户。
    事情的转机是在2016年2月，我无意在手机上看到一条信息，我的抑郁开始有救了，于是我积极联系，简单说就是我有房子，可以贷款，而且去年房价涨的离谱，可以贷很多钱。于是，我开始收拾自己的心情，当得知确定可以的时候，我忽然感觉我的病全都好了，因为我又可以承担责任，又可以养家户口了，于是过完年后我开始工作，从期权入手，无意之间接触了优矿，根据进账数目打算成立私募，改善老婆的关系，带孩子，修复和父亲的关系，更加关心妈妈，我可以确定的有几点，第一我会更低调，更稳健的发展，第二，我说的期权和50etf产品加上量化交易是我未来的发展方向。
    我今天要去办贷款手续了，回来如果成功，会发布私募招聘广告，其实我特别需要这次的免费培训机会，我想做为我新征程的起点。
    其实这篇文章不是写给别人的，是对我过去的一次告别，绝对没有半句虚言，我希望可以给我200分，满足我去上海的愿望，谢谢大家了。
    最后一篇文字理念版我会总结一下，发发联系方式和广告，未来我会继续用策略跟大家交流。
