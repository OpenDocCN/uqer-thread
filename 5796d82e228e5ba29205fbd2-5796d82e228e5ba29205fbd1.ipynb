# 一.介绍
主成分分析在统计中的地位不言自明，而因子分析像一个孪生兄弟一样，常常和主成分分析密不可分，本帖将用最简单的叙述，越过证明，只从基本的步骤来学习一下如何用python做因子分析。

因子分析研究相关阵或协方差阵的内部依赖关系，它将多个变量综合为少数几个因子，以再现原始变量和因子之间的关系。

举个例子，在资产配置时，我们常常遇到相关性比较高的资产，会带来极大的风险，我们可以把影响风险的变量分解到几个正交的因子上，以此为标准配置资产，以避免风险聚集。

# 二.正交因子模型
设$X=(X_1,\cdots,X_p)^{T}$是P个观察到的随机变量，每个变量有n个样本，$E(X)= \mu,Cov(X)= \Sigma$,又设$F=(F_1,\cdots,F_m)^T (m<p)$是m个随机变量，$E(F)=0,Cov(F)=I_m$,又设$\epsilon=(\epsilon_1,\cdots,\epsilon_p)^T$与$F$互不相关且
$$ E(\epsilon)=0,Cov(\epsilon)=D=diag(\sigma_1^2,\cdots,\sigma_p^2)[对角阵] $$
则以下模型：
$$ X_1-\mu_1=a_{11}F_1+a_{12}F_2+\cdots+a_{1m}F_m+\epsilon_1 $$
$$ X_2-\mu_2=a_{21}F_1+a_{22}F_2+\cdots+a_{2m}F_m+\epsilon_2 $$
$$ \cdots  \cdots\cdots\cdots\cdots\cdots\cdots\cdots\cdots$$
$$ X_p-\mu_p=a_{p1}F_1+a_{p2}F_2+\cdots+a_{pm}F_m+\epsilon_p $$

被称为正交因子模型。简单点，表达的方式简单点：
$$X=\mu+AF+\epsilon$$
$F_1,\cdots,F_m$称为$X$的`公共因子`，$\epsilon_1,\cdots,\epsilon_m$称为$X$的`特殊因子`，各特殊因子之间以及特殊因子和所有公共因子之间是互不相关的。

$A$是待估计的系数矩阵，称为`因子载荷阵`

由以上的模型，我们有以下协方差结构：
$$\Sigma-D=AA^T$$
$$Cov(X,F)=A$$

因子分析的目的首先是由样本协方差阵$\hat \Sigma$估计$\Sigma$,然后分解得到A和D，再来分析预测公共因子$F_1,\cdots,F_m$,因为A中元素$a_{ij}$刻画了$X_i$和$F_j$之间的相关性，称为$X_i$在$F_j$上的因子载荷。

* `共同度`：

	记A中第i行元素的平方和为$h_i^2$,$$h_i^2 = \sum_{j=1}^{m} a_{ij}^2 ,(i=1,2,\cdots,p)$$称为变量$X_i$的共同度，反映了变量$X_i$对公因子$F$依赖的程度。

* `贡献度`：
	
    记A中第j列元素的平方和为$q_j^2$,$$q_j^2=\sum_{i=1}^p a_{ij}^2,(j=1,2,\cdots,m)$$称为$F_j$对$X$的贡献，表示第j个公共因子$F_j$对$X$的所有分量$X_1,\cdots,X_p$的总影响

* 注：
	
    模型不受量纲影响。
	
	因子载荷阵A不唯一。

# 三.参数估计方法
## 1.主成分法
设样本的协方差阵S的特征值为$\lambda_1 \geq \lambda_2 \geq \lambda_3 \geq \cdots \geq \lambda_p \geq 0$,对应的单位正交向量为$l_1 , l_2 , \cdots l_p$

当最后$p-m$个特征值较小时，S可以按照$S=AA^T+D$的形式进行分解，那么模型的一个解得到如下：
$$A=(\sqrt \lambda_1 l_1,\cdots,\sqrt \lambda_m l_m)$$
$$\sigma_i^2=s_{ii}-\sum_{t=1}^m a_{it}^2 \ ,(i=1,2,\cdots,p)$$
称为因子模型的`主成分解`

$m$的确定可以根据实际问题，也可以按照前m个特征值所占权重和达到某个值为标准,如：
$$\frac{\lambda_1+\cdots+\lambda_m}{\lambda_1+\cdots+\lambda_m+\cdots+\lambda_p} \geq P_0$$
当相关变量所取单位不同时，常常对其标准化，标准化后的样本协方差阵就是原始变量的相关阵$R$，同样也用上面的方法求出主成分解

## 2.主因子法
从$R$出发，可以有一种主成分法的修正：
$$R-D=AA^T=R^*$$
称为`约相关阵`。如果已知特殊方差的初始估计,则有：
$$(h_i^*)^2=1-(\hat \sigma_i^*)^2$$
$$R^*=\begin{pmatrix}  (h_1^*)^2 & r_{12} & \cdots & r_{1p}  \\ r_{21} & (h_2^*)^2 & \cdots & r_{2p} \\ \vdots & \vdots & \cdots & \vdots \\ r_{p1} & r_{p2} & \cdots & (h_p^*)^2 \end{pmatrix}$$
然后按照之前的方法取$R^*$的特征值和特征向量，得到分解式：
$$R^*=AA^T$$
$$A=(\sqrt \lambda_1^* l_1^*,\cdots,\sqrt \lambda_m^* l_m^*)$$
令$$\hat \sigma_i^2=1-\sum_{t=1}^{m} a_{it}^2 \quad (i=1,\cdots,p)$$
则A和$D^*=diag(\sigma_1^2,\cdots,\sigma_p^2)$为因子模型的一个解，称为`主因子解`

若$h_i^2$是未知的，则用迭代的方式来求解。

## 3.极大似然法
由于假设了因子的正态分布，并用迭代法求解，所以这里不再赘述。

# 四.例子
以上三种方法中，主成分法应用较广泛，下面举个简单的例子来计算一下因子载荷阵等：

```python
import numpy as np
import pandas as pd
from math import *
```

我随便取了以下5只基金：

```python
universe=['150008','150012','512300','512330','513100']
```

再取它们从20160401到20160701的日收益率作为样本矩阵$X$：

```python
#取日收益率，以ticker名为列名
X = pd.DataFrame()
for i in universe:
    data = DataAPI.MktFunddGet(ticker=i,beginDate=u"20160401",endDate=u"20160701",field=u"preClosePrice,closePrice",pandas="1")
    X[i] = data['closePrice']/data['preClosePrice']-1
```

```python
X.head() #看看日收益率的前5行
```

```python
X.shape
```

## 1.计算样本均值，样本离差阵$E$及样本相关阵$R$：
样本均值$\bar{x}$:
$$\bar{X}=\frac{1}{n} \sum_{t=1}^{n}X_{(t)} =(\bar{x_1},\cdots,\bar{x_p})^T$$

```python
X_mean = X.mean()
print X_mean
```

样本离差阵$E$:
$$E=\sum_{t=1}^{n} (X_{(t)}-\bar X)(X_{(t)}-\bar X)^T \overset{def}{=} (e_{ij})$$

```python
E = np.mat(np.zeros((5,5)))
for i in range(len(X)):
    E += (X.iloc[i,:].reshape(5,1)-X_mean.reshape(5,1))*(X.iloc[i,:].reshape(1,5)-X_mean.reshape(1,5))
```

样本相关阵$R$:
$$R=(r_{ij}),r_{ij}=\frac{e_{ij}}{\sqrt(e_{ii}e_{jj})}$$

```python
R= np.mat(np.zeros((5,5)))
for i in range(5):
    for j in range(5):
        R[i,j] = E[i,j]/sqrt(E[i,i]*E[j,j])
```

```python
R
```

其实也可以直接使用dataframe的函数corr

```python
X.corr()
```

## 2.求R的特征值和标准化特征值向量

```python
import numpy.linalg as nlg
eig_value,eig_vector=nlg.eig(R)  #求特征值，特征向量
```

```python
eig = pd.DataFrame()
eig['names'] = X.columns
eig['eig_value'] = eig_value
eig
```

把特征值从大到小排序$\lambda_1 \geq \lambda_2 \geq \lambda_3 \geq \cdots \geq \lambda_p \geq 0$：

```python
eig.sort('eig_value',ascending=False,inplace=True)
eig
```

## 3.求因子模型的因子载荷阵$A$
先确定公共因子的个数$m$:
这里使用前m个特征值的比重大于80%的标准，即：
$$\frac{\lambda_1+\cdots+\lambda_m}{\lambda_1+\cdots+\lambda_m+\cdots+\lambda_p} \geq 0.8$$

```python
for m in range(1,6):
    if eig['eig_value'][:m].sum()/eig['eig_value'].sum() >= 0.8:
        print m
        break
```

所以公共因子是三个，然后就可以求因子载荷阵了：
$$A=(\sqrt \lambda_1 l_1,\cdots,\sqrt \lambda_m l_m)$$

```python
A=np.mat(np.zeros((5,3)))
A[:,0]=sqrt(eig_value[0])*eig_vector[:,0]
A[:,1]=sqrt(eig_value[2])*eig_vector[:,2]
A[:,2]=sqrt(eig_value[3])*eig_vector[:,3]
```

```python
pd.DataFrame(A)
```

## 4.求特殊因子方差$\hat \sigma_i^2$和$X_i$的共同度$h_i^2$的估计
$$\hat \sigma_i^2=1-\sum_{t=1}^{m} a_{it}^2 \quad (i=1,\cdots,p)$$
$$h_i^2 = \sum_{j=1}^{m} a_{ij}^2 ,(i=1,2,\cdots,p)$$

```python
h=np.zeros(5)  #共同度
D=np.mat(np.eye(5))  #特殊因子方差
for i in range(5):
    a=A[i,:]*A[i,:].T
    h[i]=a[0,0]
    D[i,i] = 1-a[0,0]
```

```python
print h
print pd.DataFrame(D).to_html()
```

## 5.对公共因子做解释
基本想法是：首先看哪些变量在这个因子中的负荷量大，然后利用专业知识讨论这些变量组合的实际含义。

```python
a=pd.DataFrame(A)
a.columns=['factor1','factor2','factor3']
a.index=['x1','x2','x3','x4','x5']
a
```

比如看因子载荷阵的第二列，除了$a_{22}$的绝对值较大外，其它相对较小，可以说F2主要反映x2的信息
