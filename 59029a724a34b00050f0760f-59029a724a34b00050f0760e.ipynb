# 期权平价套利实践

## 期权平价套利策略
---
买卖权平价关系，即Put-Call Parity，是期权市场最为经典且易于理解的著名等式关系。其不受制于任何期权定价模型，如BS模型，二叉树模型的影响而始终保持成立。投资者无需考虑波动率因素即可根据其来判断期权价格是否偏离合理价格，从而发现套利机会。

本文分析讨论了买卖期权平价关系，给出了原始平价套利策略以及变形买卖权平价关系的套利策略。

## 1. 原始平价套利策略简介
----
原始平价套利策略是基于买卖权平价关系（Put-Call Parity）而产生的。在忽略市场交易成本的情况下，相同标的资产且执行价格也相同的看涨期权、看跌期权与标的资产价格之间满足如下公式关系。其中T为到期日、r为同期限无风险收益率、并假设标的资产S不进行分红。
$$C + K×e^{-rT} = P + S$$
我们可以通过如下现金流分析来验证该等式的有效性：
<center>![](http://storage-uqer.datayes.com/56e7cf9c228e5b858bc25952/d08854b4-2bb1-11e7-9a94-0242ac140004)</center>
<br>
如上图所示，在t=T时刻，持有等式两边资产组合的价值是相同的，根据无套利定价原理，在t=0时刻两个资产组合的价值也应该相同，否则将存在无风险套利的机会。基于上述买卖权平价关系，当公式所示关系不成立时，我们可以实施平价套利策略，而该策略存在下述两个方向：
<br>
* 1.1. 做空看涨期权组合、做多看跌期权组合
<br>
当出现$C + K×e^{-rT} > P + S$的情况时，看涨期权被高估，因此我们卖出看涨期权，并借入$K×e^{-rT}$价值的无风险资产，买入低估的看跌期权，同时买入标的资产。在t=0时刻，我们获得现金净流入$CF_{0} = C + K×e^{-rT} - P - S > 0$，而在到期日t=T时刻，现金流无变化。
<br>
* 1.2. 做空看跌期权组合，做多看涨期权组合
<br>
当出现$C + K×e^{-rT} < P + S$的情况时，看跌期权被高估，因此我们卖出高估的看跌期权，同时卖出标的资产，买入低估的看涨期权，同时借出$K×e^{-rT}$价值的无风险资产。在t=0时刻，我们获得现金净流入$CF_{0} = P + S -C - K×e^{-rT} > 0$，而在到期日t=T时，现金流无变化。

`实际比较可行操作`:
我们不考虑借入借出无风险资产，使用自有资金代替。另外由于华夏上证50ETF不能进行单纯的卖空操作，因此当出现$C + K×e^{-rT} < P + S$的情形时，我们无法卖出标的资产进行套利，故我们此处仅考虑出现$C + K×e^{-rT} > P + S$的情况。

实际上，期权平价套利忽略了卖出期权的保证金占用，因为通常在海外保证金产生的利息是归期权所有人所有的，所有没有影响，但在我国这部分利息通常是被经纪商拿去了。故更简化的了可以这么看，当`做空看涨期权、做多看跌期权组合`产生的收益除以标的资产与保证金的成本得到的收益率大于无风险收益率，或者得到了收益率比较有诱惑力时，我们可以做这样一次套利，`做空看涨期权、做多看跌期权组合`。

```python
def base_opt_arbitrage_rate(Cprice,Pprice,Kprice,Sprice,expire_date_length):
    
    call_expire_return = (Cprice - max(Sprice - Kprice,0))*10000
    put_expire_return = (max(Kprice - Sprice,0) - Pprice)*10000
    expire_spot_return = 0
    deposit = get_option_deposit(Cprice,Sprice,Kprice,'C')
    porfilo_expire_return = call_expire_return + put_expire_return + expire_spot_return - 25 ## 手续费设置为25块
    arbitrage_rate = porfilo_expire_return / (Sprice*10000 + deposit)
    anunal_arbitrage_rate = (1+arbitrage_rate)**(250.0/expire_date_length) - 1
    return anunal_arbitrage_rate

def get_option_deposit(option_price,objectPrice,strikePrice,opt_type):
    """
    计算期权保证金
    """
    if opt_type == 'C':
        outMoney = strikePrice - objectPrice
        if outMoney < 0 :
            outMoney = 0
        return (option_price + max(0.12*objectPrice - outMoney,0.07*objectPrice))*10000
    elif opt_type == 'P':
        outMoney = strikePrice - objectPrice
        if outMoney > 0 :
            outMoney = 0
        return (option_price + max(0.12*objectPrice - outMoney,0.07*objectPrice))*10000
```

## 2. 变形平价套利策略
------
在本部分中讨论将该公式进行一定的变形，并形成新的等式关系。假设市场中存在相同标的资产，相同到期日、不同执行价格的两组期权合约（看涨期权及对应看跌期权）。则存在如下两个等式，其中所有期权合约到期日相同均为T，标的资产相同均为S，区别仅仅在于执行价格：
<center>$C_1 + K_1×e^{-rT} = P_1 + S$</center>
<center>$C_2 + K_2×e^{-rT} = P_2 + S$</center>
<center>$K_2 > K_1$</center>
将两个等式相减，稍作变形我们可以得到：
<center>$C_1 + P_2 + (K_1 - K_2)×e^{-rT} = P_1 + C_2$</center>
根据之前我们对于原始买卖权平价关系等式的证明，所得等式两边资产组合应保持等价，否则会出现套利机会。我们按以下两个方向讨论：

* 1.1. 做空$C_1+P_2$、做多$P_1+C_2$
<br>
当出现$C_1 + P_2 + (K_1 - K_2)×e^{-rT} > P_1 + C_2$的情况时，若套利空间可观，我们卖出$C_1$与$P_2$，买入$P_1$与$C_2$。
<br>
* 1.2. 做空$P_1+C_2$、做多$C_1+P_2$
<br>
当出现$C_1 + P_2 + (K_1 - K_2)×e^{-rT} < P_1 + C_2$的情况时，若套利空间可观，我们买入$C_1$与$P_2$，卖出$P_1$与$C_2$。

`实际操作`:
只要$C_1 + P_2 + (K_1 - K_2)×e^{-rT} \neq P_1 + C_2$,我们就可以卖出高的买入低的，锁定利润；关键是这个利润的是否足够高，高过买入卖出期权的手续费加上无风险利息，这样的收益就是比较诱人的。

```python
def expand_opt_arbitrage_rate(C1,P1,K1,C2,P2,K2,Sprice,expire_date_length):
    
    c1_price = C1*10000
    p1_price = P1*10000
    c2_price = C2*10000
    p2_price = P2*10000
    deposit_c1 = get_option_deposit(C1,Sprice,K1,'C')
    deposit_p1 = get_option_deposit(P1,Sprice,K1,'P')
    deposit_c2 = get_option_deposit(C2,Sprice,K2,'C')
    deposit_p2 = get_option_deposit(P2,Sprice,K2,'P')
    
    if c1_price + p2_price+(K2-K1)*10000*math.exp(-0.03*expire_date_length/250) > p1_price + c2_price:
        deposit = deposit_c1 + deposit_p2
        porfilo_expire_return = c1_price-max(Sprice-K1,0)*10000 + p2_price - max(K2-Sprice,0)*10000 - p1_price - c2_price + max(K1-Sprice,0)*10000 +max(Sprice-K2,0)*10000 - 50 ## 手续费设置为50
        arbitrage_rate = porfilo_expire_return / (deposit + p1_price + c2_price)
    else:
        deposit = deposit_p1 + deposit_c2
        porfilo_expire_return = p1_price - max(K1-Sprice,0)*10000 + c2_price - max(Sprice-K2,0)*10000 - c1_price - p2_price + max(Sprice-K1,0)*10000 + max(K2-Sprice,0)*10000 - 50
        arbitrage_rate = porfilo_expire_return / (deposit + c1_price + p2_price)
    anunal_arbitrage_rate = (1+arbitrage_rate)**(250.0/expire_date_length) - 1
    return anunal_arbitrage_rate
```

## `附录`

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
class OptionExpireReturnAnalysis():
    
    def __init__(self):
        pass
    
    def expire_call_return(self,Cprice,Kprice,direction=None,amount=None,S=None):
        if direction is None:
            direction = 1
        if amount is None:
            amount = 1
        if S is None:
            S = Kprice
        S_price_window = np.linspace(S/2,1.5*S,100)
        expire_return = {}
        if direction == 1:
            for s in S_price_window:
                if s > Kprice:
                    expire_return[s] = ((s - Kprice) - Cprice) * amount
                else:
                    expire_return[s] = - Cprice * amount
        elif direction == -1:
            for s in S_price_window:
                if s > Kprice:
                    expire_return[s] = ((Kprice - s) + Cprice) * amount
                else:
                    expire_return[s] = Cprice * amount
        return pd.Series(expire_return)*10000
    
    def expire_put_return(self,Pprice,Kprice,direction=None,amount=None,S=None):
        if direction is None:
            direction = 1
        if amount is None:
            amount = 1
        if S is None:
            S = Kprice
        S_price_window = np.linspace(S/2,1.5*S,100)
        expire_return = {}
        if direction == 1:
            for s in S_price_window:
                if s < Kprice:
                    expire_return[s] = ((Kprice - s) - Pprice) * amount
                else:
                    expire_return[s] = - Pprice * amount
        elif direction == -1:
            for s in S_price_window:
                if s < Kprice:
                    expire_return[s] = ((s - Kprice) + Pprice) * amount
                else:
                    expire_return[s] = Pprice * amount
        return pd.Series(expire_return)*10000
    
    def spot_expire_return(self,S,direction):
        S_price_window = np.linspace(S/2,1.5*S,100)
        expire_return = {}
        if direction == 1:
            for s in S_price_window:
                expire_return[s] = s - S
        elif direction == -1:
            for s in S_price_window:
                expire_return[s] = S - s
        return pd.Series(expire_return)*10000
    
    def expire_return_plot(self,expire_return_list,legend):
        fig = plt.figure(figsize=(8,5))
        ax = fig.add_subplot(111)
        color_list = ['b','r','g','y','pink','orange','purple']
        for i,r in enumerate(expire_return_list):
            r = r.map(lambda x:np.round(x,8))
            r.plot(ax=ax,color = color_list[i])
        # ax.spines['left'].set_position('zero')
        ax.spines['right'].set_color('none')
        ax.yaxis.tick_left()
        # ax.spines['bottom'].set_position('zero')
        ax.spines['top'].set_color('none')
        ax.xaxis.tick_bottom()
        plt.xlabel('$S_T$')
        plt.ylabel('return')
        ax.legend(legend,frameon=False,loc='lower center')
        ax.grid(True, which='both')
```
