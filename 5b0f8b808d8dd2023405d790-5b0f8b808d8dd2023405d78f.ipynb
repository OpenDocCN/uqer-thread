```python
stockCentralDF = stockDF[stockDF["partyNatureCD"]=="0101"].reset_index(drop=True)
print stockCentralDF.head(10).to_html()
print stockCentralDF.shape
```

再进一步，可以看一下这384只股票里面哪一些近两年有定增预案公布

```python
beginDate = u"20160509"
stockCentralPPDF = DataAPI.EquPrivateFloatGet(secID=u"",ticker=stockCentralDF["ticker"],beginDate=beginDate,endDate=u"",field=u"",pandas="1")
stockCentralPPDF = stockCentralPPDF.drop_duplicates(subset="ticker").reset_index(drop=True)
print stockCentralPPDF.to_html()
```

还剩32只股票了，是不是范围一下子缩小了很多，真的是选择困难症患者福音。来吧，看看等权买入这32只票的话过去两年表现如何。

```python
chgPctDF = pd.DataFrame()
for ticker in stockCentralPPDF['ticker']:
    mktDF = DataAPI.MktEqudGet(secID=u"",ticker=ticker,tradeDate=u"",beginDate=u"20160508",endDate=u"",isOpen="",field=u"",pandas="1")
    mktDF.set_index(["tradeDate"], inplace=True)
    chgPctDF = chgPctDF.append(mktDF['chgPct'].T)
chgPctDF.ix['sum'] = chgPctDF.sum()/chgPctDF.shape[0]
chgPctDF.ix[-1,0] = 0
nav_series = chgPctDF.ix['sum'].add(1.0).cumprod()
indexPrice = DataAPI.MktIdxdGet(tradeDate=u"",indexID=u"",ticker=u"000300",beginDate=beginDate,endDate=u"",exchangeCD=u"XSHE,XSHG",field=[u"tradeDate",u"closeIndex"],pandas="1")
indexPrice.index = indexPrice['tradeDate'].tolist()
benchmarkNav = indexPrice.copy()
benchmarkNav["benchmarkNav"] = 1+(benchmarkNav["closeIndex"]/benchmarkNav["closeIndex"].ix[0]-1)
```

```python
#plot
ax = (nav_series).plot(figsize=(16,6), label='nav')
ax = (benchmarkNav["benchmarkNav"]).plot(figsize=(16,6), color='grey', style='--',ax=ax, label='indexPrice')
ax.legend(loc='upper left')
plt.title('nav_series', fontproperties=font, fontsize=18)
plt.show()
```
