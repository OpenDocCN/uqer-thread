# 基于江恩理论的趋势选股

本策略主要是基于江恩理论以及Dual Thrust方法为依据。
主要策略为对市场前期股票数据做一定判断，划定相对激进（策略一）与相对保守（策略二）的两个策略在行情看好与震荡市分别进行。

策略一：
range：前30日收盘价的最大值减去前30日收盘价的最小值
上轨：前30日开盘价的平均值加上range
下轨：前30日开盘价的平均值减去range
买入信号：收盘价上穿上轨或收盘价上穿下轨
卖出信号：收盘价下穿上轨或收盘价下穿下轨
止盈条件：收益大于20%，止盈
止损条件：亏损大于6%，止损
