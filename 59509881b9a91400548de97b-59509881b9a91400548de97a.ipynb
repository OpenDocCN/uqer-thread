一个实现邮件推送功能的小程序，收集有关股票的新闻事件并发送到指定邮箱。

```python
import numpy as np
import pandas as pd
import CAL.PyCAL as cal

pd.set_option("display.max_colwidth",500) #必须要设很大值，否则加链接的标题转换成HTML格式时不会得到正确的结果！！！
# 调用CAL的日期处理功能，设定时间参数
today=cal.Date.todaysDate() #今日日期
today=str(today)

# 设置关注股票
tickerList = ['600000','000001']

# 定义获得新闻的函数
def fun_newsMessage(tickerList,beginDate,endDate):
    # 获取资讯
    df=DataAPI.NewsByTickersGet(ticker=tickerList,beginDate=beginDate,endDate=endDate,\
            field=u"ticker,secShortName,newsID,newsTitle,newsPublishTime,newsPublishSite",pandas="1")
    df0=df.drop_duplicates('newsTitle') #筛除重复新闻
    df0['newsTitle']=df0['newsTitle'].str.replace('"','*') #去除标题中的引号
    newsID=list(df0['newsID'])
    newsIDList=[str(i) for i in newsID]
    df=DataAPI.NewsBodyGet(newsID=newsIDList,field=u"newsID,newsURL",pandas="1")
    df1=pd.merge(df0,df,how='inner',on='newsID')
    df1['newsTitleURL']='<a href="'+df1['newsURL']+'" target="_blank" >'+df1['newsTitle']+'</a>'
    df1=df1[['ticker','secShortName','newsTitleURL','newsPublishSite','newsPublishTime']] #选择要显示的列并排序
    df1=df1.sort_values('ticker') #按代码排序
    df1=df1.rename(columns={'ticker':'代码','secShortName':'名称','newsTitleURL':'标题','newsPublishSite':'来源','newsPublishTime':'发布时间'}) #修改列标题名称
    message=''
    message=message+df1.to_html(escape=False) #转换成HTML，对<>不进行处理
    return message
    

# 设置邮件标题和内容
subject="今日关注["+str(today)+"]"
message=fun_newsMessage(tickerList,today,today)
print(message)


# 发送邮件
mail=send_mail('aaa@bbb.com',subject,message)
if mail['code']==200:
    print('发送成功！')
```
