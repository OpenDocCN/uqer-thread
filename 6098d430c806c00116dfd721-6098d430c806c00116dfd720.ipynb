```python
# 导入常用的模块，默认步骤
import pandas as pd
import numpy as np
import seaborn
import matplotlib.pyplot as plt
import datetime as dt
import copy
```

```python
# 抽取多因子投资的换仓日序列，类似于回测时候的参数设置
global factor_dfTotal

beginDate=u"20200301"
endDate=u"20210501"
universe="000906"
quantile_groups=5

field="calendarDate"
tradeDate_df = DataAPI.TradeCalGet(exchangeCD=u"XSHG",beginDate=beginDate,endDate=endDate,isOpen=u"1",field=field,pandas="1")
tradeDate_list=[ x for x in tradeDate_df["calendarDate"].values if dt.datetime.strptime(x,"%Y-%m-%d").isoweekday()==1]
print "换仓日序列，但是 优矿uqer 的机制使得实际下单在下一个交易日的开盘"
print tradeDate_list
```

```python
# 抽取每个交易日的因子数据
df_list=[]

for tradeDate in tradeDate_list:
    field="consTickerSymbol,consShortName,intoDate"
    universe_SecID_df=DataAPI.IdxConsGet(secID=u"",ticker=universe,isNew=u"",intoDate=tradeDate,field=field,pandas="1")
    universe_SecID_df["tradeDate"]=tradeDate
    universe_SecID_df=universe_SecID_df[["tradeDate"]+field.split(",")]
    #print universe_SecID_df

    factor_list="OperatingRevenueGrowRate,PB,LCAP".split(",")
    ticker=universe_SecID_df["consTickerSymbol"].tolist()
    field=["tradeDate","ticker","secID"]+factor_list
    factor_df=DataAPI.MktStockFactorsOneDayGet(tradeDate=tradeDate,secID=u"",ticker=ticker,field=field,pandas="1")

    for factor in factor_list:
        factor_df=factor_df.sort_values(by=factor)
        group_num=np.array([quantile_groups]*len(factor_df))
        for i in range(quantile_groups):
            group_num[int(np.floor(1.0*len(factor_df)*(i-1)/quantile_groups)):
                      int(np.floor(1.0*len(factor_df)*i/quantile_groups))]=i
        factor_df["group_"+factor]=group_num
    factor_df=factor_df.sort_values(by="ticker").reset_index(drop=True)
    df_list.append(factor_df)
    print "finish loading data: "+tradeDate

print "\n"
print "这就是某个交易日，股票池的因子数据，最标准的形式："
print factor_df[["tradeDate","ticker"]+factor_list]
```

```python
# 合并成因子数据总表
factor_dfTotal=pd.concat(df_list,axis=0).reset_index(drop=True)
print "因子数据总表"
print factor_dfTotal
```

```python
# 抽取交易日序列的价格、收益率数据，便于计算单因子IC值
df_list=[]

for tradeDate in tradeDate_list:
    df_cut=factor_dfTotal[ factor_dfTotal["tradeDate"]==tradeDate ]
    ticker_list=df_cut["ticker"].tolist()
    
    field="tradeDate,ticker,secShortName,closePrice"
    stockReturn_df=DataAPI.MktEqudGet(ticker=ticker_list,tradeDate=tradeDate,
                                      isOpen="1",field=field,pandas="1")
    df_list.append(stockReturn_df)
    print "finish loading data: "+tradeDate

print "股票的价格和收益率数据加载完毕"
#print stockReturn_df
```

```python
# 合并股票池的价格和收益率数据
stockReturn_dfTotal=pd.concat(df_list,axis=0).reset_index(drop=True)

df_list=[]
for ticker in stockReturn_dfTotal["ticker"].unique():
    df_cut=stockReturn_dfTotal[ stockReturn_dfTotal["ticker"]==ticker ].sort_values(by="tradeDate")
    df_cut["return"]=0.0
    df_cut["return"]=df_cut["closePrice"].shift(-1)/df_cut["closePrice"]-1
    df_list.append(df_cut[["return"]])

dfTotal_cut=pd.concat(df_list,axis=0)
stockReturn_dfTotal=pd.concat([stockReturn_dfTotal,dfTotal_cut],axis=1,join="inner").sort_index()

print "股票两次换仓之间收益率的总表"
print stockReturn_dfTotal
```

```python
# 计算IC值，实际上IC值就是 “秩相关系数”，其计算公式比较简单，不妨自己动手做做

def calculate_IC(factor_name,dfTotal_factor,dfTotal_stockReturn):
    # 用于保存IC的计算结果
    tradeDate_list=[]
    IC_list=[]
    
    # 从因子数据表和股票日收益率表中，抽取出每一个交易日的数据，计算这个universe股票池中,这个factor_name因子,当天（两次换仓日间的收益率）的IC值
    df_1 = dfTotal_factor[["tradeDate","ticker",factor_name]].set_index(["tradeDate","ticker"])
    df_2 = dfTotal_stockReturn[["tradeDate","ticker","return"]].set_index(["tradeDate","ticker"])
    df_merge = pd.concat([df_1,df_2],axis=1,join="inner").dropna()
    df_merge=df_merge.reset_index(level="ticker")
    
    for tradeDate in sorted(np.unique(df_merge.index.values)):
        df_cut = df_merge.loc[tradeDate,:]
        df_cut=df_cut.sort_values(by=factor_name)
        df_cut["rank_factor"]=range(len(df_cut))
        df_cut=df_cut.sort_values(by="return")
        df_cut["rank_return"]=range(len(df_cut))
        df_cut=df_cut.sort_values(by="ticker")
        
        # IC全称Information Coefficient，表示所选股票的因子值与股票下期收益率的截面相关系数
        IC_value=np.corrcoef(df_cut["rank_factor"].values,df_cut["rank_return"].values)[0,1]
        tradeDate_list.append(tradeDate)
        IC_list.append(IC_value)
        #print "finish caculating IC: "+tradeDate
        
    df_IC=pd.DataFrame()
    df_IC["tradeDate"]=tradeDate_list
    df_IC["IC"]=IC_list
    return df_IC

        
df_IC=calculate_IC("PB",factor_dfTotal,stockReturn_dfTotal)
df_IC.plot(x="tradeDate",y="IC",figsize=(16,9),title=u"IC value of the factor",kind="bar")
```

```python
"""
多个“单因子”合并为一个“合成因子”，需要确定各个单因子的权重。
而且，每个单因子的量纲大小不同，如某个股票的总市值可能是几十亿，但是它的市净率估值可能只有1.0到20.0的数值大小。
为了将多个单因子在“同一量纲大小”水平上加总到一起，就需要处理数据。
这就是 去极值、中性化、归一化的3个流程步骤。

麻烦的是，这3个步骤的细节也各种各样。像优矿Uqer、天软tinysoft都对每个步骤提供了专门的函数。
优矿上就有 winsorize \ neutralize \ standardize 这3个函数。每个函数都有自己的输入参数，来实现具体处理数据的细节差异，例如是否需要考虑行业差异、如何去除极值等等
具体的问题，可以查看这一篇帖子： https://uqer.datayes.com/v3/community/share/55ff6ce9f9f06c597265ef04

这里，我们重新自己写一遍这3个函数。采用的都是最简单直白的模式，就是让大家看清楚，其实这3个步骤很必要，但多数情况也很简单。
"""

def func_winsorize(data_list):
    data_sorted=sorted(data_list)
    index_low=int(np.floor(0.05*len(data_list)))
    index_high=int(np.ceil(0.95*len(data_list)))
    
    value_low=data_sorted[index_low]
    value_high=data_sorted[index_high]
    
    data_winsorized=data_list
    for i,value in enumerate(data_winsorized):
        if value<value_low:
            data_winsorized[i]=value_low
        elif value>value_high:
            data_winsorized[i]=value_high
    return data_winsorized

def func_neutralize(data_list):
    # 这个函数没有实现，因为它是消除因子在不同行业之间的差异。在本帖子的框架中就不做引进了，本帖子没有考虑任何估值、营业收入的因子数值，在不同行业之间的差异。
    return data_list

def func_standardize(data_list):
    mean_value=np.mean(data_list)
    std_value=np.std(data_list)
    
    data_standardized=data_list
    for i,value in enumerate(data_standardized):
        data_standardized[i]=(value-mean_value)/std_value
    return data_standardized

a=np.random.rand(200)*20+10000
a[0]=9950; a[-1]=10099;
df_figure=pd.DataFrame()
ax1=plt.axes()
df_figure["a"]=a
df_figure["a"].plot(y="a",title="data before winsorize \ neutralize \ standardize",ax=ax1,figsize=(16,9))

print "这是 winsorize \ neutralize \ standardize 3个数据处理步骤的效果："
fig=plt.figure(figsize=(16,9))
ax2=fig.add_subplot(111)
df_figure["b"]=func_standardize(func_neutralize(func_winsorize(df_figure["a"].values)))
ax2.plot(df_figure["b"].values)
ax2.set_title("data after winsorize \ neutralize \ standardize")
```
