## **1. 目的 & 适用场景**


在使用优矿进行策略研究时，经常碰到的场景是：需要对策略内部的参数进行调整优化，此时就需要不断修改参数并不断运行backtest函数。

熟悉Quartz回测框架的用户可能了解过，Quartz在回测时会先根据回测起始日期、调仓频率等获取回测需要的行情数据以供回测使用，之后才会不断触发handle_data的逻辑。

对于以上参数调优的场景来说，每一次运行backtest函数都要获取一次同样的数据显然会造成性能损失。为了解决这一问题，优矿提供了quick_backtest函数，可以让用户快速完成批量的回测。

其具体用法如下：

```python
start = '2015-12-30'                       # 回测起始时间
end = '2016-11-01'                         # 回测结束时间
benchmark = 'HS300'                        # 策略参考标准
universe = DynamicUniverse('HS300')        # 证券池，支持股票和基金
capital_base = 1000000                     # 起始资金
freq = 'd'                                 # 策略类型，'d'表示日间策略使用日线回测，'m'表示日内策略使用分钟线回测
refresh_rate = Monthly(1)                  # 调仓频率，表示执行handle_data的时间间隔，若freq = 'd'时间间隔的单位为交易日，若freq = 'm'时间间隔为分钟

sim_params = quartz.SimulationParameters(start, end, benchmark, universe, capital_base, max_history_window=5)
data = quartz.get_backtest_data(sim_params)   #  获取回测区间所用基础行情数据

def initialize(account):                   # 初始化虚拟账户状态
    pass

def handle_data(account):                  # 每个交易日的买入卖出指令
    pass

strategy = quartz.TradingStrategy(initialize, handle_data)

# 回测部分，将data显式传入
bt, perf = quartz.quick_backtest(sim_params, strategy, data=data, refresh_rate=refresh_rate)

out_keys = ['annualized_return', 'volatility', 'information_ratio', 'sharpe', 'max_drawdown', 'alpha', 'beta']
for key in out_keys:
    print key, perf[key]
```

通过以上代码可以发现，只要调整的参数不包括start、end、benchmark、universe、capital_base、max_history_window（因为需要依赖这些参数预先获取数据），quick_backtest()函数都可以帮助我们完成批量回测的性能优化。

下面我们通过两个具体的例子来展示quick_backtest用法：

## **2. 参数选择：双均线策略** 

一个简单的对上证50择时的双均线策略思路如下：
- 以上证50ETF为标的（510050.XSHG），每天收盘后计算MA5、MA20
- 若MA5 > MA20，且当前不持仓该股，买入；反之，MA5 <= MA20，且当前持仓该股，卖出

接下来，我们要优化长短两个窗口的长度，我们使用list of tuple params_list变量进行记录，其中tuple的第一个元素为短均线长度、第二个元素为长均线长度。

```python
import pandas as pd
start = '2011-01-01'                       # 回测起始时间
end = '2016-11-01'                         # 回测结束时间
benchmark = 'SH50'                        # 策略参考标准
universe = ['510050.XSHG']        # 证券池，支持股票和基金
capital_base = 10000                     # 起始资金
freq = 'd'                                 # 策略类型，'d'表示日间策略使用日线回测，'m'表示日内策略使用分钟线回测
refresh_rate = 1                 # 调仓频率，表示执行handle_data的时间间隔，若freq = 'd'时间间隔的单位为交易日，若freq = 'm'时间间隔为分钟

sim_params = quartz.SimulationParameters(start, end, benchmark, universe, capital_base, max_history_window=60)

data = quartz.get_backtest_data(sim_params)   #  获取回测区间所用基础行情数据

params_list = [(5,15), (5,10), (10,20), (10,60), (15,60)]

for min_win, max_win in params_list:
    def initialize(account):                   # 初始化虚拟账户状态
        pass

    def handle_data(account):                  # 每个交易日的买入卖出指令
        hist = account.get_attribute_history('closePrice', max_win)
        for stk in account.universe:
            short_ma = hist[stk][-min_win:].mean()
            long_ma = hist[stk][:].mean()

            if short_ma > long_ma and stk not in account.security_position:
                order_pct_to(stk, 1.0)
            elif short_ma < long_ma and stk in account.security_position:
                order_to(stk, 0)

    strategy = quartz.TradingStrategy(initialize, handle_data)

    # 回测部分，将data显式传入
    bt, perf = quartz.quick_backtest(sim_params, strategy, data=data, refresh_rate=refresh_rate, max_history_window=60)

    out_keys = ['annualized_return', 'volatility', 'information_ratio', 'sharpe', 'max_drawdown', 'alpha', 'beta']
    print "short window length is ", min_win
    print "short window length is ", max_win
    print pd.Series(perf)[out_keys]
    print "==================="
```

## **3. 因子回测：价值/估值类因子** 

优矿为大家提供了大量的量化因子，这些因子哪些在长期表现稳定可以产生稳定获利的策略呢？

这时，用户就需要对于大量的因子进行批量的测试。下面的例子，我们演示使用quick_backtest在价值/估值类因子批量测试时的应用

```python
start = '2010-12-30'                       # 回测起始时间
end = '2016-11-01'                         # 回测结束时间
benchmark = 'ZZ500'                        # 策略参考标准
universe = DynamicUniverse('HS300') + DynamicUniverse('ZZ500')  # 证券池，支持股票和基金
capital_base = 1000000                      # 起始资金
freq = 'd'                                 # 策略类型，'d'表示日间策略使用日线回测，'m'表示日内策略使用分钟线回测
refresh_rate = Monthly(-1)                          # 调仓频率，表示执行handle_data的时间间隔，若freq = 'd'时间间隔的单位为交易日，若freq = 'm'时间间隔为分钟

sim_params = quartz.SimulationParameters(start, end, benchmark, universe, capital_base, max_history_window=5)
data = quartz.get_backtest_data(sim_params)

factor_list = ["PE", "PB", "PS", "LCAP", "LFLO", "TA2EV", "SGRO", "NLSIZE"]
factor_config = {'PE': 0, 'PB': 0, 'PS': 0, 'PCF': 0, 'ASSI': 0, 'LCAP': 0, "LFLO": 0, "TA2EV": 0, "PEG3Y": 0, "PEG5Y": 0, "PEIndu": 0, "PBIndu": 0, "PSIndu": 0, "PCFIndu": 0, 
                 "PEHist20": 0, "PEHist60": 0, "PEHist120": 0, "PEHist250": 0, "StaticPE": 0, "ForwardPE": 0, "TotalAssets": 0, "MktValue": 0, "NegMktValue": 0, "TEAP": 0, "NIAP": 1, 
                 "CETOP": 1, "SGRO": 1, "NLSIZE": 0}  # 1表示顺序，0表示逆序

# factor_list = ["PE", "PB", "PS", "PCF", "ASSI", "LCAP", "LFLO", "TA2EV", "PEG3Y", "PEG5Y", "PEIndu", "PBIndu", "PSIndu", "PCFIndu", "PEHist20", "PEHist60", "PEHist120", "PEHist250",
#               "StaticPE", "ForwardPE", "TotalAssets", "MktValue", "NegMktValue", "TEAP", "NIAP", "CETOP", "SGRO", "NLSIZE"]
# factor_config = {'PE': 0, 'PB': 0, 'PS': 0, 'PCF': 0, 'ASSI': 0, 'LCAP': 0, "LFLO": 0, "TA2EV": 0, "PEG3Y": 0, "PEG5Y": 0, "PEIndu": 0, "PBIndu": 0, "PSIndu": 0, "PCFIndu": 0, 
#                  "PEHist20": 0, "PEHist60": 0, "PEHist120": 0, "PEHist250": 0, "StaticPE": 0, "ForwardPE": 0, "TotalAssets": 0, "MktValue": 0, "NegMktValue": 0, "TEAP": 0, "NIAP": 1, 
#                  "CETOP": 1, "SGRO": 1, "NLSIZE": 0}

bt_record = dict()
risk_record = dict()

for factor_name in factor_list:
    print "Backtesting Factor : ", factor_name
    
    def initialize(account):                   # 初始化虚拟账户状态
        account.factor_name = factor_name

    def handle_data(account):                  # 每个交易日的买入卖出指令
        factor_data = DataAPI.MktStockFactorsOneDayProGet(tradeDate=account.previous_date.strftime("%Y%m%d"),secID=account.universe,field=u"secID,%s" % account.factor_name,pandas="1")
        factor_data = factor_data.set_index('secID')
        # print factor_data
        if not factor_config[account.factor_name]:
            input_data = 1.0/factor_data["%s" % account.factor_name]
        else:
            input_data = factor_data["%s" % account.factor_name]
         
        try:
            factor_data = standardize(neutralize(winsorize(input_data), account.previous_date.strftime("%Y%m%d"), exclude_style_list=['BETA', 'RESVOL', 'MOMENTUM', 'SIZE', 'SIZENL', 'EARNYILD', 'BTOP', 'GROWTH', 'LEVERAGE', 'LIQUIDTY']))
            buy_list = long_only(factor_data.to_dict(), 1, 0.2, 3, account.previous_date.strftime("%Y%m%d"))
        except Exception, e:
            print e
            # print "factor data ", factor_data
            print "DO NOT REBALANCE on %s" % account.current_date.strftime("%Y%m%d")
            return

        my_univ = buy_list.keys()

        univ = [x for x in my_univ if x in account.universe]

        # 不在股票池中的，清仓
        for stk in account.security_position:
            if stk not in univ:
                order_to(stk, 0)
                
        # 在目标股票池中，按比例买入
        for stk in univ:
            if 0 <= buy_list[stk] <= 1:
                order_pct_to(stk, buy_list[stk])
    
    strategy = quartz.TradingStrategy(initialize, handle_data)
    # 回测部分
    try:
        bt, perf = quartz.quick_backtest(sim_params, strategy, data=data, refresh_rate=refresh_rate)

        bt_record[factor_name] = bt
        risk_ret = dict()
        out_keys = ['annualized_return', 'volatility', 'information_ratio', 'sharpe', 'max_drawdown', 'alpha', 'beta']
        for key in out_keys:
            risk_ret[key] = perf[key]
            risk_record[factor_name] = risk_ret
    except Exception, e:
        print e
        continue
```
