##一、函数
    函数是组织好的，可重复使用的，用来实现单一，或相关联功能的代码段。Python中，函数就是带函数名的代码块，它能提高应用的模块性，和代码的重复利用率。函数分为内建函数和自定义函数。
    Python中的函数，无论是命名函数，还是匿名函数，都是语句和表达式的集合。在Python中，函数是第一个类对象，这意味着函数的用法并没有限制。Python函数的使用方式就像Python中其他值一样，例如字符串和数字等。Python函数拥有一些属性，通过使用Python内置函数dir就能查看这些属性。
    1. __doc__返回指定函数的文档字符串。
    2. __name__返回函数名字。
    3. __module__返回函数定义所在模块的名字。
    4. func_defaults返回一个包含默认参数值的元组
    5. func_globals返回一个包含函数全局变量的字典引用。
    6.  func_dict返回支持任意函数属性的命名空间

```python
def foo():
    """这是一个函数"""
    for i in range(11):
        print i
print dir(foo)
print foo.__class__
print foo.__doc__
print foo.__name__
print foo.__module__
print foo.__format__
```

###1.1定义函数
下面我们来定义一个函数，定义函数有一些规则：
    1、函数代码块以def关键词开头，后接函数标识符名称和圆括号()。
    2、任何传入参数和自变量必须放在圆括号中间。圆括号之间可以用于定义参数。
    3、函数的第一行语句可以选择性地使用文档字符串—用于存放函数说明。
    4、函数内容以冒号起始，并且缩进。
定义函数的语法：
    def 函数名（[参数列表]）： ##参数列表可选
         函数体
###1.2调用函数
定义一个函数只给了函数一个名称，指定了函数里包含的参数，和代码块结构。这个函数的基本结构完成以后，你可以通过另一个函数调用执行，也可以直接从Python提示符执行。

```python
def handle_data(data):
    data1 = []
    for i in data:
        if i % 2 == 0:
            data1.append(i)
    return data1
l = range(11)
print handle_data(l)
```

###1.3 函数参数
函数参数分为形参和实参。形参是函数定义时传入的参数，没有实际意义，而实参是函数调用传入的参数。像上文的data就是形参，而l就是实参。
Python函数可以使用的参数类型：1.必备参数，2.命名参数，3.缺省参数，4.不定长参数

1、必备参数

必备参数须以正确的顺序传入函数。调用时的数量必须和声明时的一样。

```python
def foo1(name,age):
    print "hello, %s.The age is %d"%(name,age) 
foo1("Mike",23)##这就是一个必备参数，参数顺序很重要。
```

2、命名参数

命名参数和函数调用关系紧密，调用方用参数的命名确定传入的参数值。你可以跳过不传的参数或者乱序传参，因为Python解释器能够用参数名匹配参数值。

```python
def foo1(name,age):
    print "hello, %s.The age is %d"%(name,age) ##这就是一个必备参数，参数顺序很重要。
foo1(age=23,name="Mike") ##这就是一个命名参数，不用考虑参数顺序。
```

3、缺省参数

调用函数时，缺省参数的值如果没有传入，则被认为是默认值。

```python
def foo1(name,age = 18):
    print "hello, %s.The age is %d"%(name,age) ##这就是一个必备参数，参数顺序很重要。
foo1("Mike",23) 
foo1("Mary")##这就是一个缺省参数，年龄默认18。
```

4、不定长参数

你可能需要一个函数能处理比当初声明时更多的参数。这些参数叫做不定长参数，和上述2种参数不同，声明时不会命名。

```python
def foo2(*args,**kwargs):
    print "args=",args
    print "kwargs=",kwargs
foo2(1,2,3,4)
foo2(a=1,b=2,c=3)
foo2(1,2,3,4, a=1,b=2,c=3)
foo2('a', 1, None, a=1, b='2', c=3)
```

可以看到，这两个是python中的可变参数。*args表示任何多个无名参数，它是一个tuple；**kwargs表示关键字参数，它是一个dict。并且同时使用*args和**kwargs时，必须*args参数列要在**kwargs前。
