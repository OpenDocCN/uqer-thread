### 2  DataCube

一个接口就能同时获取`行情、财务、因子`等多资产多维度的数据，研究更加便捷。
下面的示例，展示了如何使用股票的行情数据，计算指标或者因子。 

```python
symbols = ['600000.XSHG']
fields = ['high', 'low']
start ='2016-01-01'
end = '2016-04-18'

data = get_data_cube(symbols, fields, start, end, style='sat')
for symbol in symbols:
    current = data[symbol]        
    # 计算指标或因子 Average True Range
    current['ATR'] = current['high'] - current['low']
    print current.head().to_html()
```

## 研究

### 1 代码补全
Notebook中，单元格为『代码』 或『策略』模式时，输入部分变量或函数，按 Tab 键后，会进行代码自动补全，若有多个选项时，显示补全列表。
输入代码的时候，按tab键，则会显示代码补全提示以及代码帮助，按enter键则会自动补全。
上述enter键补全代码后，`如果是函数，则会提示该函数的参数列表`。
光标放在函数的括号内，按`ctrl+i`，则会提示该函数的参数列表。

### 2 debug调试功能
在Notebook中，您可以对策略和代码单元进行调试。

您也可以点击代码行号处，对要进行调试的代码行设置调试断点，再点击『调试』进入代码调试模式。

### 6 风险模型(`Pro`)
优矿风险模型共提供了以下9个数据接口：
1.	因子暴露数据 DataAPI.RMExposureDayGet
2.	因子收益数据 DataAPI.RMFactorRetDayGet
3.	特质收益数据 DataAPI.RMSpecificRetDayGet
4.	风险因子协方差矩阵表(日级别) DataAPI.RMCovarianceDayGet
5.	风险因子协方差矩阵表(short类型) DataAPI.RMCovarianceShortGet
6.	风险因子协方差矩阵表(long类型) DataAPI.RMCovarianceLongGet
7.	特质风险表(day类型) DataAPI.RMSriskDayGet
8.	特质风险表(short类型) DataAPI.RMSriskShortGet
9.	特质风险表(long类型) DataAPI.RMSriskLongGet

专业版用户可以直接在优矿的Notebook中调用，也可以通过优矿的SDK在本地获取。

### 13 重启研究环境
重启研究环境是优矿为用户提供的深度使用功能。您可能会遇到Notebook开启过多不愿意一个个关闭，在小概率情况下研究环境启动异常等情况，此时您可以选择使用“重启研究环境”功能。

您可以在资源管理器中点击“重启研究环境”按钮，使用该功能将导致您所有的Notebook都重启，您当前运行的notebook变量都将在内存中清除，但是Notebook文件中显示的内容不受影响，请谨慎使用
