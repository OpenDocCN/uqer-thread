```python
import numpy as np
import pandas as pd


import cPickle as pickle
import datetime

#通联库
from CAL.PyCAL import *
cal = Calendar('China.SSE')

import numpy as np
import pandas as pd

import scipy
import math


#Ours
from lib.data_loader import load_market_data
# from lib.Alpha101Factor import alpha101_fun
from lib.Quality101Factor import  quality101_fun
def Winsorize(x):
    inplace =True
    limits  =0.01
    return  x.apply(scipy.stats.mstats.winsorize)

def zscore(x):
    mu = np.nanmean(x)
    sigma =np.nanstd(x)
    x = (x - mu) / sigma;  
    return x

def quantile_calc(series, n):
    edges = pd.Series([float(i) / n for i in range(n + 1)])
    f = lambda x: (edges >= x).argmax()
    return series.rank(pct=1).apply(f)

# 获取某日股票行业对应关系
def get_ticker_sector(securities,current_date):
    ticker_sector_df = DataAPI.EquIndustryGet(secID=securities, industryVersionCD='010303', intoDate=current_date, field='secID,secShortName,industry,industryID1,industryName1')
    ticker_sector_df['industryName1'] = [x.decode('utf8') for x in ticker_sector_df['industryName1']]
    sector_mappings = ticker_sector_df[['secID','industryName1']].copy()
    sector_mappings.set_index('secID',inplace=True)
    sector_mappings = sector_mappings['industryName1'].to_dict()
    return sector_mappings
```
