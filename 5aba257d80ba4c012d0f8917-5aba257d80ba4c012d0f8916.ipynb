```python
# 本代码目的：获取PB，PE，PS等因子,按时间存入矩阵，并绘制分位数图，直观了解市场或板块的当前估值状态

import numpy as np
import pandas as pd
import datetime
import matplotlib.pyplot as plt
from CAL.PyCAL import *
cal = Calendar('China.SSE')
#------------------------------------------------------------------------------------------------------------------
stk_list=set_universe('A')
data1=DataAPI.EquIndustryGet(industry=u"",secID=stk_list,industryVersionCD=u"010303",intoDate=u"20180326",industryID1=u"",industryID2=u"",industryID3=u"",field=u"secID,secShortName,industryName1,industryName2,industryName3",pandas="1")
bank=data1[data1.industryName3=='银行']
uni=list(bank.secID)
print bank
for i0 in bank.index[20:]:
    name=bank.secShortName[i0]
    stk=bank.secID[i0]
# bl=[['600016.XSHG','民生银行']]
# bl.append(['600015.XSHG','华夏银行'])
# bl.append(['601328.XSHG','交通银行'])
# bl.append(['601988.XSHG','中国银行'])
# bl.append(['601818.XSHG','光大银行'])
# for key in bl:
    # stk,name=key
    print name,stk
    begin = datetime.date(2007,1,1)                  # 设定起始时间
    end = datetime.date(2018,3,23)                    # 设定终止时间
    Qt=['PB','PB_BankMedian','PE','PE_BankMedian','PS','PS_BankMedian','EPS','EPS_BankMedian','ROA','ROA_BankMedian','ROA5','ROA5_BankMedian','ROE','ROE_BankMedian','ROE5','ROE5_BankMedian','BLEV','BLEV_BankMedian','DEGM','DEGM_BankMedian','NPParentCompanyGrowRate','NPParentCompanyGrowRate_BankMedian','OperatingProfitRatio','OperatingProfitRatio_BankMedian','OperatingRevenueGrowRate','OperatingRevenueGrowRate_BankMedian','close','turnoverRate','turnoverRate_BankMedian']
    #-----------------------------------------------------------------------------------------------------------------------
    Tdays = cal.bizDatesList(begin, end)             # 得到起止时间之间的交易日列表
    data = pd.DataFrame(index = Tdays, columns = Qt)
    for d in Tdays:                          # 遍历Tdays时间列表中的日期
        try:
            f_str='PB'
            for f_str in ['PB','PE','PS','EPS','ROA','ROA5','ROE','ROE5','BLEV','DEGM','NPParentCompanyGrowRate','OperatingProfitRatio','OperatingRevenueGrowRate']:
                df0 = DataAPI.MktStockFactorsOneDayGet(tradeDate=d,secID=uni,field=u"secID,"+f_str).set_index("secID")
                data[f_str][d]=df0[f_str][stk]
                data[f_str+'_BankMedian'][d]=df0.dropna().quantile([0.5]).values[0][0]
            #----------------------------------------------------------------------------------------------------------------
            for f_str in ['turnoverRate',]:
                df0 = DataAPI.MktEqudGet(secID=uni,tradeDate=d,field=u"secID,"+f_str).set_index("secID")
                data[f_str][d]=df0[f_str][stk]
                data[f_str+'_BankMedian'][d]=df0.dropna().quantile([0.5]).values[0][0]
            close=DataAPI.MktEqudGet(secID=stk,tradeDate=d,field=u"closePrice,secID").set_index("secID").values[0][0]   
            data['close'][d]=close
        except:
            pass
    data.to_csv('./银行详细因子/'+name+'_PB等.csv')
```
