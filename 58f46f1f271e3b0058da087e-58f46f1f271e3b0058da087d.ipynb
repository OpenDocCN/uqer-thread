```python
import numpy as np
import pandas as pd
import talib as ta
import datetime
from dateutil.parser import parse#修改时间格式的作用
import matplotlib.pyplot as plt
from CAL.PyCAL import *    # CAL.PyCAL中包含font
font.set_size(15)
ticker='000001'
begin='20100103'
end='20161213'
M=30
```

```python
#获取指数数据
df_index = DataAPI.MktIdxdGet(ticker=ticker,beginDate=begin,endDate=end,field=u"tradeDate,preCloseIndex,closeIndex",pandas="1")
df_index['index_ret'] = df_index['closeIndex'] / df_index['preCloseIndex'] - 1
df_index['tradeDate'] = df_index['tradeDate'].map(lambda x:parse(x)) 
```

```python
def get_trade_calendar(start_date,end_date):
    """获取交易日历列表"""
    trade_date = DataAPI.TradeCalGet(exchangeCD=u"XSHG",beginDate=start_date,endDate=end_date,field=u"isOpen,calendarDate",pandas="1")
    trade_calendar = trade_date[trade_date['isOpen']==1]
    trade_calendar['calendarDate'] = trade_calendar['calendarDate'].map(lambda x:x.replace('-',''))#修改日期格式 注意
    trade_calendar = trade_calendar['calendarDate'].tolist()
    return trade_calendar
def get_component_return(index_symbol,start_date,end_date):
    """获取动态股票池行情"""
    trade_calendar = get_trade_calendar(start_date,end_date)
    df_eq = pd.DataFrame()
    for date in trade_calendar:#循环
        universe = set_universe(index_symbol,date)#动态股票池
        df_tmp = DataAPI.MktEqudAdjAfGet(secID=universe,tradeDate = date,
                                field=u"secID,tradeDate,preClosePrice,closePrice,openPrice",pandas="1")
        df_eq = pd.concat([df_eq,df_tmp],axis=0)#dataframe 的合并方式 添加到下面的作用
    return df_eq
```

```python
trade_calendar = get_trade_calendar(begin,end)
df_eq = get_component_return(ticker,begin,end) #是一个很长的  
```

```python
df_eq
```

```python
df_eq['eq_ret'] = df_eq['closePrice'] / df_eq['preClosePrice'] - 1
df_eq['eq_ret'] = df_eq.apply(lambda x: np.NaN if x['openPrice']==0 or np.isnan(x['openPrice']) else x['eq_ret'],axis=1) #获取个股收益率，停牌则设为缺失值      openPrice 是0的 或者是空值的  axis=1 是按列查看  if else 语句
```

```python
df_eq_ret = df_eq[['secID','tradeDate','eq_ret']]#可以直接这么写 
df_eq_ret = df_eq_ret.set_index(['tradeDate','secID'])#一定要注意 这个写法 可以提两个 index  那么他会把一个重复的合并 
eq_ret = df_eq_ret.unstack()['eq_ret'] # 获取个股历史收益率   注意格式
eq_ret.index=pd.to_datetime(eq_ret.index, format='%Y-%m-%d')#调日期格式  index一定要对上 否则合不起来
df_index=df_index.set_index(df_index['tradeDate'])
df_index.index = pd.to_datetime(df_index.index, format='%Y-%m-%d')
data = pd.concat([eq_ret,df_index['index_ret']],axis=1)#和大盘涨跌幅
data['ii'] = range(len(data))#加一列  方便下面有pd rolling apply
```

```python
data #获取了所有的数据 大盘的 个股的 每天的收益率  下面就是作回归了
```

```python
def rsquare(x,y):
    ##计算拟合优度R2
    meanofx=x.mean()
    meanofy=y.mean()
    difofxx=x - meanofx
    difofxy=x - meanofy
    difofyy=y - meanofy 
    beta = sum(difofxx * difofyy) / sum(difofxx ** 2)
    r2 = beta**2 * sum(difofxx**2) / sum(difofyy**2)
    return r2
def window_convergence(x,data):
    # 计算趋同度
    window_data =  data.iloc[map(int,x)]#  是一种函数的执行  把x都转换成int 
    window_data = window_data.dropna(axis=1)
    r2_list = []
    for col in window_data.columns:
        if col != 'index_ret':
            r2_list.append(rsquare(window_data['index_ret'],window_data[col]))
    return np.nanmean(r2_list)
```

```python
convergence_30 = pd.rolling_apply(data.ii, M, lambda x: window_convergence(x, data))#这应该是一种循环格式
#每次x 都是一个list 是[0到29] [1到30]。。。。 这种写法应该多注意
convergence_30.to_csv('convergence_30_shangzheng.csv')
```
