# `Pandas`

```python
from pandas import Series, DataFrame
import pandas as pd
import numpy as np
```

## `Series`
---

A Series is a one-dimensional array-like object containing an array of data and an associated array of data labels.  The data can be any NumPy data type and the labels are the Series' index.

```python
print 'Create a Series:'

ser_1 = Series([1, 1, 2, -3, -5, 8, 13])
print ser_1

print '\nGet the array representation of a Series:'
print ser_1.values

# Index objects are immutable and hold the axis labels and metadata such as names and axis names.

print '\nGet the index of the Series:'
print ser_1.index

print '\nGet values great than 0:'
print ser_1[ser_1 > 0]

print '\nScalar multiply:'
print ser_1 * 2
```

## `DataFrame`

A DataFrame is a tabular data structure containing an ordered collection of columns.  Each column can have a different type.  DataFrames have both row and column indices and is analogous to a dict of Series.  Row and column operations are treated roughly symmetrically.  Columns returned when indexing a DataFrame are views of the underlying data, not a copy.  To obtain a copy, use the Series' copy method.


```python
print '\nCreate a DataFrame:'

data_1 = {'state' : ['VA', 'VA', 'VA', 'MD', 'MD'],
          'year' : [2012, 2013, 2014, 2014, 2015],
          'pop' : [5.0, 5.1, 5.2, 4.0, 4.1]}
df_1 = DataFrame(data_1)
print df_1.to_html()

print '\nCreate a DataFrame specifying a sequence of columns:'
df_2 = DataFrame(data_1, columns=['year', 'state', 'pop'])
print df_2.to_html()

print '\nRetrieve a column by key, returning a Series:'
print df_2['year']

print '\nRetrieve a row by position:'
print df_2.ix[0]

print '''\nAssign a new column that doesn't exist to create a new column:'''
df_2['new_column'] = range(len(df_2))
print df_2.to_html()

print '\nTranspose the DataFrame:'
print df_2.T.to_html()

print '\nReturn the data contained in a DataFrame as a 2D ndarray:'
print df_2.values
```

```python
print '\nDrop rows from a Series or DataFrame:'
df_3 = df_2.drop([0, 1])
print df_3.to_html()

print '\nDrop columns from a DataFrame:'
df_3 = df_3.drop('new_column', axis=1)
print df_3.to_html()

print '\nSelect a slice from a DataFrame:'
print df_2[:1].to_html()

print '\nSelect from a DataFrame based on a filter:'
print df_2[df_2.year > 2014].to_html()

print '\nSelect a slice of rows from a DataFrame (note the end point is inclusive):'
print df_2.ix[2:3].to_html()

print '\nSelect a slice of rows from a specific column of a DataFrame:'
print df_2.ix[0:2, 'year']

print '\nSelect rows based on an arithmetic operation on a specific row:'
print df_2.ix[df_2.year > 2014].to_html()
```

```python
print '\nArithmetic and Data Alignment'
np.random.seed(0)
df_4 = DataFrame(np.random.rand(9).reshape((3, 3)),
                 columns=['a', 'b', 'c'])
print df_4.to_html()

print '\nApply a function on 1D arrays to each column:'
func_1 = lambda x: x.max() - x.min()
print df_4.apply(func_1)

print '\nApply a function on 1D arrays to each row:'
print df_4.apply(func_1, axis=1)

print '\nApply a function and return a DataFrame:'
func_2 = lambda x: Series([x.min(), x.max()], index=['min', 'max'])
print df_4.apply(func_2).to_html()

print '\nApply an element-wise Python function to a DataFrame:'
func_3 = lambda x: '%.2f' %x
print df_4.applymap(func_3).to_html()
```

```python
print '\nSummarizing and Computing Descriptive Statistics:'
print df_4.describe()

print '\nsum values:'
print df_4.sum()

print '\nsum values on each row:'
print df_4.sum(axis=1)
```

```python
print '\nCleaning Data :'
data_1 = {'state' : ['VA', 'VA', 'VA', 'MD', 'MD'],
          'year' : [2012, 2013, 2014, 2014, 2015],
          'population' : [5.0, 5.1, 5.2, 4.0, 4.1]}
df_5 = DataFrame(data_1)
print df_5.to_html()

print '\nReplace all occurrences of a string with another string, in place (no copy):'
df_5.replace('VA', 'VIRGINIA', inplace=True)
print df_5.to_html()

print '''\nDrop the 'population' column and return a copy of the DataFrame:'''
df_6 = df_5.drop('population', axis=1)
print df_6.to_html()

print '\nConcatenate two DataFrames:'
data_2 = {'state' : ['NY', 'NY', 'NY', 'FL', 'FL'],
          'year' : [2012, 2013, 2014, 2014, 2015],
          'population' : [6.0, 6.1, 6.2, 3.0, 3.1]}
df_7 = DataFrame(data_2)
df_8 = pd.concat([df_5, df_7])
print df_8.to_html()
```
