# 【 大师系列 】 价值选股法 —— 史蒂夫·路佛

#  *史蒂夫·路佛价值选股法则*

* `史蒂夫·路佛（1938—）`华尔街著名投资人

>1. 任职投资经理（1969-1981）

>  * 路佛于1969-1977 年间担任Piper,Jaffray & Hopwood 公司的投资策略
分析师，并于1977-1981 年间担任德州Criterion Investment Management 公司的投
资组合经理。

>2. 成立路佛威登资产管理公司（1981-）

>  * 路佛于1981 年创建路佛集团，一个国际化的投资研究公司。1987 年开
始，路佛集团集团开始利用内部的研究成果开展证券投资业务，正式成为资产管理公
司。至2001 年3 月22 日止，路佛核心投资基金以5 年、3 年、1 年及3 个月报酬率
皆打敗S&P500 指数的优异表现，获得晨星公司(Morningstar Inc.) 美国本土混合型
基金类(Domestic Hybrid)五颗星的评价。至2005 年，公司管理的资产规模已经达到
10 亿美元。

>3. 史蒂夫·路佛的著作

>  * 路佛已经从事投资研究40 年，其1980 年出版的《The Myths
of inflation and Investing》是投资史上的经典。

> > `选股条件` ：


>     1.   市净率低于全市场平均值 （`PB`）

>     2.   以五年平均盈余计算的PE低于全市场平均值 (`ETP5`)  

>     3.   股价现金流量比低于全市场平均值(`PCF`)

>     4.   股息收益率不低于全市场平均值(`派息水平`)

>     5.   流动比率高于全市场平均值(`衡量财务状况`)

>     6.   每股流动资产至少是股价的30% (`衡量估值水平`)

>     7.   长期借款占总资本比率低于50%(`衡量财务状况`)
