```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
def is_buy(stock, previous_date):
    std = StandardScaler()
    data1=DataAPI.MktEquFlowOrderGet(secID=stock,ticker=u"",beginDate=u"20100823",endDate=previous_date,
                                     field=u"",pandas="1")
    data1 = pd.DataFrame(data1)
    data2=DataAPI.MktEqudGet(secID=stock,ticker=u"",tradeDate=u"",beginDate=u"20100823",endDate=previous_date,
                             isOpen="",field=u"",pandas="1")
    data2 = pd.DataFrame(data2)

    data2 = data2[['tradeDate', 'closePrice']]
    data = pd.merge(data1,data2,on='tradeDate')
    
    del data['secID']
    del data['ticker']
    del data['secShortName']
    
    data.set_index('tradeDate', inplace=True)
    data.index = pd.to_datetime(data.index)
    
    x_train = data[:]
    y_train = x_train[['closePrice']]

    y_train['5_closePrice'] = y_train['closePrice'].shift(-5)
    
    y_train.dropna(inplace=True)
    y_train['signal'] = np.where((y_train['5_closePrice']-y_train['closePrice']) > 0, 1, -1)
    x_train = x_train[:-5]
    x_train = x_train.fillna(method='ffill') 
   
    
    # 标准化，训练
    std = StandardScaler()
    x_train = std.fit_transform(x_train)
    lg = LogisticRegression(C=1.0)  #
    
    lg.fit(x_train, y_train['signal'])
    
    log.info(previous_date)
    
    data3=DataAPI.MktEquFlowOrderGet(secID=stock,ticker=u"",beginDate=previous_date,endDate=previous_date,
                                     field=u"",pandas="1")
    
    data3 = pd.DataFrame(data3)
    data4=DataAPI.MktEqudGet(secID=stock,ticker=u"",tradeDate=previous_date,beginDate=u"",endDate=u"",
                             isOpen="",field=u"",pandas="1")
    data4 = pd.DataFrame(data4)
    
    data4 = data4[['tradeDate', 'closePrice']]
    data5 = pd.merge(data3,data4,on='tradeDate')
    
    
    del data5['secID']
    del data5['ticker']
    del data5['secShortName']
    
    data5.set_index('tradeDate', inplace=True)
    data5.index = pd.to_datetime(data5.index)
    
    
    x_test = data5
    # x_test = x_test.fillna(method='ffill')
    
    
    x_test = std.transform(x_test)

    y_predict = lg.predict(x_test)
    log.info(y_predict)
#     print(y_predict)
    # print(y_predict[0])
    if(y_predict[0]==1):
        return True
    else:
        return False


#     x_test = data[2001:]
#     x_test = x_test.fillna(method='ffill') 
#     y_test = x_test[['closePrice']]
#     y_test['5_closePrice'] = y_test['closePrice'].shift(-5)

#     y_test['signal'] = np.where((y_test['5_closePrice']-y_test['closePrice']) > 0, 1, -1)

#     y_test.dropna(inplace=True)
#     x_test = x_test[:-5]
#     x_test = std.transform(x_test)
#     predict = lg.pre

    
    


    
    
```
