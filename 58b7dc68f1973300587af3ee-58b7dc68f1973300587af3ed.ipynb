```python
import numpy as np
from CAL.PyCAL import *    # CAL.PyCAL中包含font
from datetime import timedelta,datetime
import datetime
import time
import scipy.stats as st
cal=Calendar('China.SSE')
import random
```

```python
def buylist(input_account_date,universe0):
    calDate = Date.fromDateTime(input_account_date)
    xgend=cal.advanceDate(calDate,'-1B',BizDayConvention.Preceding)
    df=DataAPI.MktEqumGet(secID=universe0,monthEndDate=xgend,field=u"secID,chgPct,tradeDays",pandas="1")
    df=df[df.tradeDays>df.tradeDays.mean()*0.8]
    buylistall=df.sort(['chgPct'],ascending=True)[:20].secID.tolist()
    return buylistall
```
