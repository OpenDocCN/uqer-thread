# 使用板块历史PE区间带进行估值分析

做投资，无法避免的要面对估值的问题，方法很多，最常用的就是PE、PEG估值。 

对于某个行业的历史PE区间带，能够让你对当前股票价格处于一个什么样的估值区间，能有一个客观的认识。

当然PE低不是你买入的理由， PE高也未必是卖出的理由。有些行业PE低，是对风险的补偿，有些行业PE高，是包含了对未来的预期。

当市场泡沫在快速膨胀的时候，PE也随之水涨船高， 此时对历史PE极值的了解，能够让你在别人狂热的时候，保持有一颗清醒的心。 有了这颗心，根据“钟摆理论”，当趋势发生改变的时候，在别人迷茫、恐惧的时候，你能够比别人多一分清醒，或许逃命的时候，你就比别人快一些。

当中国南车，达到"交通运输"板块2010年的极值90倍PE的时候，股价明显有阻力。

对于大多数股票，PE是随着股价波动,  但是估值不能单纯只考虑，还要根据行业特点，进行分类讨论。
比如资源类的股票，往往在大宗商品价格高昂的时候，公司收入增加，具有较低PE,  而在大宗商品低迷的时候，公司收入减少，具有较高的PE。  
比如互联网类的股票，往往无法用PE进行估值。
比如消费类的股票，往往业绩比较稳定，股价和PE成正相关。

在日常交易中，估值仅仅是决定买卖的因子之一。比如相对较低的估值，会给你一个向下的安全边际，但是如果股票质地有问题，而股票是可以长时间保持低估值。 极端不合理的高估值，能给你随时准备沽空的底气。 
而最终决定买卖，需要靠多个因子发生共振，形成合力。

对于中长线投资者来说，PE区间带的低点区域和高点区域，比中间区域更具有参考意义。比如最近几年在PE区间带的低点， 15倍PE买入上证50， 25倍PE买入沪深300， 基本上不会有大的亏损。

最后,从指数估值的角度来衡量一下目前点位的估值：
上证50，  45倍左右的PE， 高于本轮牛市之前30倍的PE。
沪深300， 30倍左右的PE， 略高于本轮主板牛市之前的20-30倍PE。
中证500， 58倍左右的PE， 高于本轮牛市之前的30-50倍PE。
创业板，  100倍左右的PE， 相当于本轮主板牛市之前的估值，而创业板从2013年开始走牛，当时的起点是40倍PE。

市场是否足够便宜，吸引到价值投资者筑底，还需要仁者见仁智者见智。
