```python
# coding:utf-8
# ===============================================================
# 名称：YF-工具-深度下跌底部抢反弹V0.2
# 功能：超跌反弹选股简化模型，寻找短期过度下跌有抢反弹机会的股票
# 时间：2017-12-11
# 策略概述：算法是给定时间区间、下跌幅度
# 和底部范围，找出A股市场内同时满足以下二个条件的个股以供进一步分析。
#     条件一：股价超跌：最小值在最大值之后并且（最大值-最小值）/最大值>=给定的下跌幅度；
#     条件二：当前股价处于底部范围：当前股价<=最小值*底部范围阀值
# 策略优缺点：
#     优点：采用前复权数据，下跌幅度计算更真实；可以迅速给出A股市场所有符合条件的备选股
#     缺点：
#       1) 观察区间内上市的次新股会被忽略；
#       2）区间内配送除权的股票可能会被误选；
#       3）未考虑这些股票的成交量、基本面和消息面情况，需人工加以分析辨别；
# 不能直接进行交易。
# 更新日志：
# 1）V0.1版实现了价格变化的分析
# ================================================================

# 开启缓存，当前Notebook所有DataAPI数据都会在内存进行缓存
DataAPI.settings.cache_enabled = True #调试时设为False

# 导入模块
import numpy as np
import pandas as pd
import CAL.PyCAL as cal

# 设定条件
SPAN=cal.Period('6M') #默认设定时间区间为半年，如3个月3M，1年12M
DOWN_RATE=0.5 #默认下跌幅度50%
BOTTOM_RANGE=0.1  #距离最低点不超过10%视为底部


# 调用CAL的日期处理功能，设定时间参数
today=cal.Date.todaysDate() #今日日期
calendar=cal.Calendar('China.SSE')  #建立A股市场日历对象
end=calendar.adjustDate(today, cal.BizDayConvention.Preceding) #时间区间结束日期
begin=calendar.adjustDate(today-SPAN, cal.BizDayConvention.Preceding) #时间区间开始日期

# print(len(calendar.bizDatesList(begin,end))) #显示观察区间的交易日数

universe=set_universe('A') #获取A股股票代码

#获取股票代码和简称
df0=DataAPI.EquGet(equTypeCD=u"A",secID=universe,listStatusCD=u"",field=u"ticker,secShortName",pandas="1")

#获取行情信息
df=DataAPI.MktEqudGet(secID=universe,beginDate=begin,endDate=end,isOpen="",
                      field=u"ticker,tradeDate,closePrice,accumAdjFactor,turnoverVol",pandas="1")
df['adjBefore']=df.closePrice*df.accumAdjFactor  #计算前复权收盘价
df=df[['ticker','tradeDate','adjBefore']]
df.columns=['ticker','tradeDate','closePrice']  #收盘价是前复权收盘价
ticker=df[['ticker']].drop_duplicates()['ticker'].tolist()
df_tmp=pd.DataFrame()
for i in ticker:
    df1=df[df.ticker==i]
    df_now=df1.tail(1)
    df_now=df_now[['ticker','closePrice']]
    df1=df1.sort_values('closePrice')
    df2=pd.merge(df1.tail(1),df1.head(1),how='inner',on='ticker')
    df2=pd.merge(df2,df_now,how='inner',on='ticker')
    df_tmp=pd.concat([df_tmp,df2])
df_tmp.columns=['ticker','highDate','highPrice','lowDate','lowPrice','currPrice']
df_tmp=df_tmp[df_tmp.highDate<df_tmp.lowDate]
df_tmp['downRate']=(df_tmp.highPrice-df_tmp.lowPrice)/df_tmp.highPrice
df_tmp=df_tmp[(df_tmp.currPrice<=df_tmp.lowPrice*(1+BOTTOM_RANGE))&(df_tmp.downRate>=DOWN_RATE)]
df_tmp=pd.merge(df0,df_tmp,how='inner',on='ticker')
df_tmp=df_tmp.sort_values('downRate',ascending=False)
df_tmp.columns=['股票代码','股票名称','高点日期','高点价格','低点日期','低点价格','现价','下跌幅度']
print('超跌反弹选股简化模型：')
df_tmp

```
