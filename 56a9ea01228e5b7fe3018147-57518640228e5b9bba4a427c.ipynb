```python
def Orgr5(account,k):
    today = account.current_date
    cal = Calendar('China.SSE')
    lastDay = cal.advanceDate(today,'-1B',BizDayConvention.Unadjusted)
    today_str = lastDay.strftime("%Y%m%d")
    start_date = cal.advanceDate(lastDay,'-6Y',BizDayConvention.Unadjusted)
    start_date_str = start_date.strftime("%Y%m%d")
    equity = pd.DataFrame()
    for rpt_type in ['A']:
        try:
            tmp = DataAPI.FdmtISGet(secID=account.universe, field=['secID', 'endDate', 'publishDate', 'tRevenue','NIncome'], beginDate=start_date_str, publishDateEnd=today_str,  reportType=rpt_type)
        except:
            tmp = DataFrame()
        equity = pd.concat([equity, tmp], axis=0)
    equity = equity.sort(columns=['secID','endDate'], ascending=[True, False])
    equity = equity.drop_duplicates(cols=['secID','endDate'])
    equity['tRevenue5'] = equity['tRevenue'].shift(-4)
    equity['RevenueRate'] = equity['tRevenue']/equity['tRevenue5']
    equity['NIncome5'] = equity['NIncome'].shift(-4)
    equity['NIncomeRate'] = equity['NIncome']/equity['NIncome5']
    equity = equity.drop_duplicates(cols=['secID'])
    equity = equity.dropna()
    RRlist =  list(equity[equity['RevenueRate'] > k*equity.median()['RevenueRate']].secID)
    NRlist =  list(equity[equity['NIncomeRate'] > k*equity.median()['NIncomeRate']].secID)
 
    return RRlist ,NRlist 
```

```python
def Cash(account,k):
    today = account.current_date
    cal = Calendar('China.SSE')
    lastDay = cal.advanceDate(today,'-1B',BizDayConvention.Unadjusted)
    today_str = lastDay.strftime("%Y%m%d")
    start_date = cal.advanceDate(lastDay,'-6Y',BizDayConvention.Unadjusted)
    start_date_str = start_date.strftime("%Y%m%d")
    equity = pd.DataFrame()
    for rpt_type in ['A']:
        try:
            tmp = DataAPI.FdmtCFGet(secID=account.universe, field=['secID', 'endDate', 'publishDate', 'NCEEndBal'], beginDate=start_date_str, publishDateEnd=today_str,  reportType=rpt_type)
        except:
            tmp = DataFrame()
        equity = pd.concat([equity, tmp], axis=0)
    equity = equity.sort(columns=['secID','endDate'], ascending=[True, False])
    equity = equity.drop_duplicates(cols=['secID','endDate'])
    equity['NCEEndBal5'] = equity['NCEEndBal'].shift(-4)
    equity['NCEEndBalRate'] = equity['NCEEndBal']/equity['NCEEndBal5']
    equity = equity.drop_duplicates(cols=['secID'])
    equity = equity.dropna()
    CSlist =  list(equity[equity['NCEEndBalRate'] > k*equity.median()['NCEEndBalRate']].secID)   
    return CSlist
```
