# 转债不同评级等权

```python
from datetime import datetime
import pandas as pd
def get_bond(tradeDate):
    # 获取某日可转债指数所有成分
    bond = DataAPI.IdxConsGet(ticker=u"000832",intoDate=tradeDate,field=u"consID",pandas="1")
    consID = bond['consID'] 
    tmp_ID=list(consID.values)
    #去掉可交换债
    #cb = DataAPI.BondGet(secID=tmp_ID,typeID=u"02020113",field=u"secID",pandas="1")
    #bond= cb['secID']
    bond = consID.values
    return list(bond) 

cal=DataAPI.TradeCalGet(exchangeCD=u"XSHG",beginDate=u"20220101",endDate=u"20220701",isOpen=u"1",field=u"",pandas="1")
cal = cal.loc[cal['isWeekEnd']==1]
cal.index = range(len(cal))
print(len(cal))
#'CCC','B-','B','B+','BBB-','BBB','A-',
ratings = ['A','A+','AA-','AA','AA+','AAA']

def getCurrentResult(lastDate, currentDate, result, to_hold):
    #获取本月收益
    if len(to_hold)>0:
        price=DataAPI.MktBonddGet(secID=to_hold,beginDate=lastDate,endDate=currentDate,\
                                  field=u"secID,tradeDate,openPrice,closePrice",pandas="1")
        price=price.groupby(['secID','tradeDate']).mean().unstack().T
        openPrice=price.loc['openPrice'].iloc[0]
        closePrice=price.loc['closePrice'].iloc[-2]#调仓日前一天卖出
        #print openPrice,closePrice
        mean = (closePrice/openPrice-1).mean()
        if 2 > mean > -1:
            mean = mean
        else:
            mean = 0
        result[currentDate] = mean    
    else:
        result[currentDate] = 0

def singleRate(rate):
    print(rate)
    result={}
    to_hold=[]
    for i in range(len(cal)):
        if i!=0:
            getCurrentResult(cal.loc[i-1]['calendarDate'],cal.loc[i]['calendarDate'], result, to_hold)
        else:
            result[cal.loc[i]['calendarDate']] = 0
        bond=get_bond(cal.loc[i]['calendarDate'])#获取所有可转债
        #获取可转债
        z = DataAPI.BondRatingGet(secID=bond, pandas="1")
        z.drop_duplicates(subset="secID", keep="last", inplace=True)
        z = z.loc[z['rating'] == rate]
        to_hold = list(z['secID']) 
        print(cal.loc[i]['calendarDate'],len(to_hold),to_hold[:5])
    return result
idx = list(cal['calendarDate'])
clm = ratings
df = pd.DataFrame(0,index=idx,columns=clm)
for rate in ratings:
    result = singleRate(rate)
    data = (pd.Series(result)+1).cumprod()
    df[rate] = data
print(df)
df.plot()
```
