# 基金收益率分解

## 基金收益率分解的意义
FOF的本质是基金的组合，所以对于FOF产品的设计最后还是要到归结到基金的选择上。这里尝试对基金收益来源进行分解，探究基金业绩的内部因素，分离基金的择风格能力和选股能力，尝试发现基金业绩背后的稳定特质。如果能够对基金收益进行有效的分解，后面就可以进行FOF的构建。限于篇幅，本次仅仅做基金收益率分解，下篇帖子再实现FOF的构建与回测。

-----

## 算法说明
公募基金都是每季度末公布一次前10大重仓股，半年才会发布一份完整的持仓数据，这个频率是很低的。私募基金更是很难获取持仓数据。但是公募基金是可以拿到每日的收益数据的。因此这里尝试对基金的收益进行回归分析，获取基金的组合成分。通过收益分析还有另外一个好处，就是针对混合型基金可以分解其资产配置比例。

William F. Sharpe于1992年结合资产因子模型（[Asset class factor model](https://web.stanford.edu/~wfsharpe/art/sa/sa.htm)）提出收益率分析法，将资产风格划分为大盘价值、大盘成长、中盘、小盘以及不同类债券和海外市场，利用多元线性回归方法，对基金收益率进行回归，以最小化残差平方和为目标，得到基金在各资产风格上的近似比例。

<center>$R_{t}=[\delta_{1}x_{1,t}+\delta_{2}x_{2,t}+\cdots+\delta_{n}x_{n,t}] + \varepsilon_{t}$</center>
<center>$s.t. \delta_{1} + \delta_{2}+\cdots+\delta_{n}=1$</center>
<center>$\delta_{i}\geqslant 0, i=1,2,\dots,n$</center>


其中$R_{t}$代表基金t期收益率，$x_{i,t}$代表资产风格i在t期的收益率，回归系数代表基金在各资产风格上的近似配置比例。
在得到基金于各资产风格上的近似比例后，我们可以利用此比例构建基金的“历史风格指数”$FundStyleIndex_{t}$。在选定市场基准指数$Benchmark_{t}$(我们这里选择的是HS300)后我们可以得到基金在考察期任意区间内超额收益率的一种分解形式：
<center>$R_{t}-Benchmark_{t} = [R_{t}-FundStyleIndex_{t}]+[FundStyleIndex_{t}-Benchmark_{t}]$</center>

前面就代表`选股alpha`，或者简单的说就是基金管理人带来的alpha。后面一部分就是风格相较于基准的alpha,被定义为`风格alpha`。

我们主要关心的还是第一部分`选股alpha`.

1. 由于我们这边中债的指数数据从2013年12月17号才有数据(近期历史数据就会上线)，因此我们归因的时间范围要在这个时间点以后了。

3. 在使用HS300成长和HS300价值两个指数时，发现了一些值得质疑的地方。因为二者的相关系数特别高。于是我在[中证指数官网](http://www.csindex.com.cn/sseportal/csiportal/zs/jbxx/report.do?code=000919&&subdir=1)上查询了这两个指数的编制方式,发现两个指数的成份股会存在很多的交集，这个无疑增加了回归中的共线性。

在Sharpe的[Asset class factor model](https://web.stanford.edu/~wfsharpe/art/sa/sa.htm)中也选择了成长和价值，但是根据论文中的表述：
> Stocks with high book-to-price ratios are placed in the "value" stock index; the remainder are in the "growth" stock index.

Sharpe选择的基准的成份股是没有交集的。如果真的要区分成长和价值，用**HS300价值指数**和**HS300相对价值指数**可能更为合理。

所以直接通过字面意思来选择HS300成长和HS300价值这两个指数，个人觉得并不是特别合理，后面我直接用`HS300`和`ZZ500`取代。虽然HS300和ZZ500相关性也挺高，但是成份股没有交集。


当然有可能我的理解有偏差，欢迎一起讨论。

## 4. 无约束的回归
首先使用statsmodels的OLS回归，不加约束。
1. 保留HS300成长和HS300价值。
2. 先取2014年6月的收益序列做回归

通过下面的回归，可以看到R方的值比较大，但是P值也特别大，所以并不显著，应该是多重共线性导致。

```python
x_df = pd.DataFrame()

x_df['zz_13'] = zz_13_se 
x_df['zz_57'] = zz_57_se
x_df['zz_10'] = zz_10_se

x_df['HS300_grouth'] = HS300_grouth_se
x_df['HS300_value'] = HS300_value_se
x_df['zz_1000'] = zz_1000_se
x_df['zz_500'] = zz_500_se

x_df['risk_free_rate'] = r007_se.mean() / 360.0 / 100.0
x_df = x_df.dropna().ix[1:]

target_ret_se = target_ret_se.dropna()
```

```python
import statsmodels.api as sm

X = x_df.values
Y = target_ret_se.values
result = sm.OLS(Y,X).fit()
print result.summary()
```

## 5. 加上约束的回归
加上约束后，这个回归问题其实就变成一个带约束的优化问题了（当然回归本身也包含优化）。这里不太适合OLS了，直接用scipy的优化器optimize。

这里面就要手动计算一下R方，最终R方稍微降低了一点，为0.94，还是相当可观的。

因为我们测试的是2014年6月的基金数据。得到的结果是债券占比0.21，股票占比0.78，现金为0.


```python
from scipy.optimize import minimize

A = np.array(x_df.values)  # 相关指数
Y = np.reshape(np.array(target_ret_se.values),(-1,1))  # 被测基金收益

data_len = x_df.shape[1]

def func(x):
    bd = np.dot(A, x) - Y.T  
    return np.sum(bd ** 2)

cons = ({'type': 'eq', 'fun': lambda x:  1 - sum(x)})
bnds = tuple((0,1) for _ in range(data_len))
start_pos = np.ones(data_len) *(1.0 / data_len) #bounds=bnds,

options = {'disp': False, 'maxiter': 500, 'ftol': 1e-20} 
res = minimize(func, start_pos, bounds=bnds, constraints=cons, method='SLSQP',options=options)
print res.x
```

```python
A = np.array(x_df.values)  # 相关指数
Y = np.reshape(np.array(target_ret_se.values),(-1,1))  # 被测基金收益

cof = np.reshape(np.array(res.x),(-1,1))
print "R-squared:",1.0 - (np.var(Y-A.dot(cof)) / np.var(Y))
```

## 6. 下面回测20140101至20170430的风格变换
我们将HS300成长和HS300价值直接用HS300取代。

```python
def get_target_and_index_ret(start,end,target_fund_se):
    """获取基准指数收益的dataframe和目标基金的日收益序列
    """
    # 债券风格基准
    zz_13_se = DataAPI.MktIdxdGet(tradeDate=u"",indexID=u"",ticker=u"CBA00321",beginDate=start,endDate=end,exchangeCD=u"",field=u"tradeDate,CHGPct",pandas="1").set_index('tradeDate')['CHGPct'] # 中债总1-3年财富


    zz_57_se = DataAPI.MktIdxdGet(tradeDate=u"",indexID=u"",ticker=u"CBA00341",beginDate=start,endDate=end,exchangeCD=u"",field=u"tradeDate,CHGPct",pandas="1").set_index('tradeDate')['CHGPct'] # 中债总5-7年财富

    zz_10_se = DataAPI.MktIdxdGet(tradeDate=u"",indexID=u"",ticker=u"CBA00361",beginDate=start,endDate=end,exchangeCD=u"",field=u"tradeDate,CHGPct",pandas="1").set_index('tradeDate')['CHGPct']  # 中债总10年以上财富


    # 股票风格基准
    HS300_se = DataAPI.MktIdxdGet(tradeDate=u"",indexID=u"",ticker=u"000300",beginDate=start,endDate=end,exchangeCD=u"",field=u"tradeDate,CHGPct",pandas="1").set_index('tradeDate')['CHGPct']  # 沪深300

    zz_500_se = DataAPI.MktIdxdGet(tradeDate=u"",indexID=u"",ticker=u"000905",beginDate=start,endDate=end,exchangeCD=u"",field=u"tradeDate,CHGPct",pandas="1").set_index('tradeDate')['CHGPct']  #中证500
    
    zz_1000_se = DataAPI.MktIdxdGet(tradeDate=u"",indexID=u"",ticker=u"000852",beginDate=start,endDate=end,exchangeCD=u"",field=u"tradeDate,CHGPct",pandas="1").set_index('tradeDate')['CHGPct']  #中证1000

    # 现金，这里使用R007作为无风险收益的基准
    r007_se = DataAPI.MktRepodGet(secID=u"",ticker=u"131801",beginDate=start,endDate=end,field=u"tradeDate,closeRate",pandas="1").set_index('tradeDate')['closeRate']
    
    x_df = pd.DataFrame()

    x_df['zz_13'] = zz_13_se 
    x_df['zz_57'] = zz_57_se
    x_df['zz_10'] = zz_10_se

    x_df['HS300'] = HS300_se
    x_df['zz_500'] = zz_500_se
    x_df['zz_1000'] = zz_1000_se

    x_df['risk_free_rate'] = r007_se.mean() / 360.0 / 100.0
    x_df = x_df.dropna()
 
    target_ret_se =  target_fund_se[(start<=target_fund_se.index) & (target_fund_se.index<=end)]  # 华夏成长混合
    target_ret_se = target_ret_se.dropna()
    return x_df, target_ret_se



def get_optimize_cof(x_df, target_ret_se):
    """返回优化后的系数和R方"""
    A = np.array(x_df.values)  # 基准指数
    Y = np.reshape(np.array(target_ret_se.values),(-1,1))  # 被测基金收益

    data_len = x_df.shape[1]

    def func(x):
        bd = np.dot(A, x) - Y.T  
        return np.sum(bd ** 2)

    cons = ({'type': 'eq', 'fun': lambda x:  1 - sum(x)})
    bnds = tuple((0,1) for _ in range(data_len))
    start_pos = np.ones(data_len) *(1.0 / data_len) 

    options = {'disp': False, 'maxiter': 500, 'ftol': 1e-20} 
    res = minimize(func, start_pos, bounds=bnds, constraints=cons, method='SLSQP',options=options)
    cof = np.reshape(np.array(res.x),(-1,1))
    
    fund_style_ret = A.dot(cof)
    r_sq =  1.0 - (np.var(Y-fund_style_ret) / np.var(Y))    
    return res.x, r_sq, fund_style_ret.flatten()
```

```python
beginDate = '20140101'
endDate = '20170430'
date_list = pd.date_range(beginDate, endDate, freq='M').date

target_fund = DataAPI.FundNavGet( ticker='000001',beginDate=beginDate,endDate=endDate,field='ticker,endDate,ADJUST_NAV') 
target_fund['pre_nav'] = target_fund['ADJUST_NAV'].shift(1)
target_fund['fund_ret'] = target_fund['ADJUST_NAV'] / target_fund['pre_nav'] - 1.0
target_fund_se = target_fund.set_index('endDate')['fund_ret']

cof_dict = {}
r_square_dict = {}
style_alpha_dict = {}
select_alpha_dict = {}

for date in date_list:
    start = str(date.replace(day=1))
    end = str(date)
    
    x_df, target_ret_se = get_target_and_index_ret(start, end, target_fund_se)

    common_index = x_df.index.intersection(target_ret_se.index)
    x_df = x_df[x_df.index.isin(common_index)]
    target_ret_se = target_ret_se[common_index]
    
    cof, r_square,fund_style_ret = get_optimize_cof(x_df, target_ret_se)
    
    cof_dict[date] = cof
    r_square_dict[date] = r_square
    
    fund_style = pd.Series(fund_style_ret,index=target_ret_se.index)
    select_alpha_dict[date] = np.sum(target_ret_se-fund_style)
    style_alpha_dict[date] = np.sum(fund_style-x_df['HS300'])


```
