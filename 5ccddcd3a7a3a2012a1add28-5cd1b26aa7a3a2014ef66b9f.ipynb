* **多次从上证50随机选股结果统计**

```python
import random
import pandas as pd
import numpy as np

start = '2010-08-01'                     # 回测起始时间
end = '2019-4-30'                       # 回测结束时间

benchmark = 'SH50'                      # 策略参考基准
universe = DynamicUniverse('SH50')
freq = 'd'                               # 'd'表示使用日频率回测，'m'表示使用分钟频率回测
refresh_rate = 1                         # 执行handle_data的时间间隔
capital_base=1000000

accounts = {
    'stock_account': AccountConfig(account_type='security', capital_base=1000000)
}

# 调用quick_backtest方法
# a、回测初始化配置，在代码模式进行策略回测时使用
sim_params = quartz.SimulationParameters(start, end, benchmark, universe, capital_base, refresh_rate=refresh_rate, accounts=accounts)
# b、获取回测行情数据
data = quartz.get_backtest_data(sim_params)
# c、运行结果
results = {}

n = 51
for i in range(1,n):

    def initialize(context):                 # 初始化策略运行环境
        pass

    def handle_data(context):                # 核心策略逻辑
        current_universe1 = context.get_universe(exclude_halt=True)
        current_universe = random.sample(current_universe1,10)

        stock_account = context.get_account('stock_account')
        current_positions = stock_account.get_positions(exclude_halt=True)    

        today = context.now
        if today.month == 10:
            for stock in current_universe:
                price = context.current_price(stock)
                if stock not in current_positions:
                    amount = stock_account.cash / len(current_universe) / price
                    stock_account.order(stock, amount)
        elif today.month == 5:
            stock_account.close_all_positions()

    # 生成策略对象
    strategy = quartz.TradingStrategy(initialize, handle_data)

    # 开始回测, 返回说明见：https://uqer.datayes.com/help/api/
    bt, perf = quartz.quick_backtest(sim_params, strategy, data=data)

    # 数据存入字典
    results[i] = {'max_drawdown': perf['max_drawdown'], 'sharpe': perf['sharpe'], 'alpha': perf['alpha'], 'beta': perf['beta'], 'information_ratio': perf['information_ratio'], 'annualized_return': perf['annualized_return'],'benchmark_annualized_return': perf['benchmark_annualized_return'],'volatility': perf['volatility'], 'bt': bt}    

    # print "\r当前进度: {:.2%}".format(i*1./(n-1))
```

```python
results_pd = pd.DataFrame(results).T.sort_index()
results_pd.to_csv('result')
results_pd = results_pd[[u'alpha', u'beta', u'information_ratio', u'sharpe', u'annualized_return', u'benchmark_annualized_return',u'volatility',u'max_drawdown']]

results_pd.columns = [u'Alpha', u'Beta', u'信息比率', u'夏普比率',  u'年化收益',u'基准年化收益', u'策略年化波动率',u'最大回撤']
results_pd.index.name = u'times'
print results_pd.to_html()
print results_pd.mean()
```

2、时间选择
==

```python
import math

start = '2010-08-01'                     # 回测起始时间
end = '2019-4-30'                       # 回测结束时间

universe = ['000001.ZICN']                 # 证券池，支持股票和基金
benchmark = '000001.ZICN'                        # 策略参考标准
freq = 'd'                                 # 策略类型，'d'表示日间策略使用日线回测，'m'表示日内策略使用分钟线回测
refresh_rate = 1                           # 调仓频率，表示执行handle_data的时间间隔，若freq = 'd'时间间隔的单位为交易日，若freq = 'm'时间间隔为分钟

capital_base=1000000
accounts = {
    'index_account': AccountConfig(account_type='index', capital_base=1000000)
}

# 调用quick_backtest方法
# a、回测初始化配置，在代码模式进行策略回测时使用
sim_params = quartz.SimulationParameters(start, end, benchmark, universe, capital_base, refresh_rate=refresh_rate, accounts=accounts)
# b、获取回测行情数据
data = quartz.get_backtest_data(sim_params)
# c、运行结果
results = {}

def daterange(month=7):
    a = range(1,13)
    a.reverse()
    b = []
    for n in a:
        if n + month <= 12:
            b.append(n + month)
        else:
            b.append(n + month - 12)
    daterange = zip(a,b)
    return daterange

for i,j in daterange(7):
    def initialize(context):                   # 初始化虚拟账户状态 
        pass

    def handle_data(context):                  # 每个交易日的买入卖出指令
        account = context.get_account('index_account')
        prevPrice = context.history(attribute='closePrice', timeRange=1, freq=freq)
        today = context.now

        for stock in context.get_universe():
            price = prevPrice[stock]['closePrice'][0]
            if today.month == i:
                account.order(stock, 100,'open')
            elif today.month == j:
                account.close_all_positions()

        return

    # 生成策略对象
    strategy = quartz.TradingStrategy(initialize, handle_data)

    # 开始回测
    bt, perf = quartz.quick_backtest(sim_params, strategy, data=data)

    # 数据存入字典
    results[i] = {'max_drawdown': perf['max_drawdown'], 'sharpe': perf['sharpe'], 'alpha': perf['alpha'], 'beta': perf['beta'], 'information_ratio': perf['information_ratio'], 'annualized_return': perf['annualized_return'],'benchmark_annualized_return': perf['benchmark_annualized_return'], 'volatility': perf['volatility'],'bt': bt}    

    # print "\r当前进度: {:.2%}".format(1 - (i-1)/12.0)
```

```python
results_pd = pd.DataFrame(results).T.sort_index()
results_pd.to_csv('result_daterange_000001')
results_pd = results_pd[[u'alpha', u'beta', u'information_ratio', u'sharpe', u'annualized_return', u'benchmark_annualized_return',u'volatility',u'max_drawdown']]

results_pd.columns = [u'Alpha', u'Beta', u'信息比率', u'夏普比率',  u'年化收益',u'基准年化收益', u'策略年化波动率',u'最大回撤']
results_pd.index.name = u'times'
print results_pd.to_html()
```
