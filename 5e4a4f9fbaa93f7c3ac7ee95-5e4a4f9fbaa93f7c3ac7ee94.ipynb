```python
def textday(T,endday):
    tradedaylist=DataAPI.TradeCalGet(exchangeCD=u"XSHG,XSHE",beginDate=u"",endDate=u"",isOpen=u"1",field=u"",pandas="1")
    tradedaylist=tradedaylist[tradedaylist['isOpen']==1]
    tradedaylist=tradedaylist[tradedaylist.calendarDate>'2015-01-01']
    tradedaylist=tradedaylist[tradedaylist.calendarDate<'2019-12-31']
            # tradedaylist['month']=tradedaylist.calendarDate.apply(lambda x:x[5:7])
            # tradedaylist['year']=tradedaylist.calendarDate.apply(lambda x:x[:4])
            # tradedaylist.drop_duplicates(subset=['month','year'],keep='first')
    t_date=tradedaylist['calendarDate'].values
            #t_date=[datetime.datetime.strptime(x,'%Y-%m-%d') for x in t_date]
            #t_date1=[datetime.datetime.strftime(x,'%Y-%m-%d') for x in t_date]
    t_date1=t_date.tolist()
    a=t_date1.index(endday)
    b=a-T
    return t_date1[b:a]
```

```python
def begin_day(T,endday):
    tradedaylist=DataAPI.TradeCalGet(exchangeCD=u"XSHG,XSHE",beginDate=u"",endDate=u"",isOpen=u"1",field=u"",pandas="1")
    tradedaylist=tradedaylist[tradedaylist['isOpen']==1]
    tradedaylist=tradedaylist[tradedaylist.calendarDate>'2015-01-01']
    tradedaylist=tradedaylist[tradedaylist.calendarDate<'2019-12-31']
    t_date=tradedaylist['calendarDate'].values
    t_date1=t_date.tolist()
    a=t_date1.index(endday)
    b=a-T
    return t_date1[b]
```

```python
def ConDegree(index,factorname,conf,T,endday): 
    
    beginday=begin_day(T,endday)
    
    ##############################################
    # 提取从beginday到endday的股票池index中的股票数据
    ##############################################
    
    stock_data=DataAPI.MktEqudGet(secID=index,beginDate=beginday,endDate=endday
                                  ,field=['secID','closePrice','tradeDate'],pandas="1").set_index('secID').dropna()
    stock_data['Rerate']=stock_data[stock_data['tradeDate']==endday]['closePrice']/stock_data[stock_data['tradeDate']==beginday]['closePrice']-1
    stock_data=stock_data[stock_data['tradeDate']==endday]
    stock_data=stock_data.dropna()
    
    ####################
    # 计算前后20%的收益差
    ####################
    
    different=np.mean((stock_data.sort(columns='Rerate').tail(int(len(stock_data)/5))['Rerate'])-np.mean(stock_data.sort(columns='Rerate').head(int(len(stock_data)/5))['Rerate'])) 
    
    #################################################
    # 提取beginday的股票池index中的因子factorname的数据
    #################################################
    
    factor_data=DataAPI.MktStockFactorsOneDayGet(tradeDate=beginday,secID=stock_data.index
                            ,field=['secID',factorname],pandas='1').set_index('secID')
    if(len(factor_data.dropna())<1):
        factor_data = factor_data.fillna(1)
    stock_data[factorname]=standardize(neutralize(winsorize(factor_data[factorname])
                                                 ,endday.replace('-',''), industry_type='SW2')).dropna()
    
    ########################
    # 计算因子前后20%的收益差
    ########################
    
    if(conf):
        different2=np.mean(stock_data.sort(columns=factorname).tail(int(len(stock_data)/5))['Rerate'].dropna())-np.mean(stock_data.sort(columns=factorname).head(int(len(stock_data)/5))['Rerate'].dropna())
    else:
         different2=np.mean(stock_data.sort(columns=factorname).head(int(len(stock_data)/5))['Rerate'].dropna())-np.mean(stock_data.sort(columns=factorname).tail(int(len(stock_data)/5))['Rerate'].dropna()) 
            
    ########################
    # 得出贡献度
    ########################
    
    return different2/different
```

```python
def Gxdtext(derr1,derr2,derr3,derr4,derr5): #,derr2,derr3,derr4,derr5
    a1=[]
    a2=[]
    a3=[]
    a4=[]
    a5=[]
    for x in c:
        ruslt1=ConDegree(index_zz500,derr1,0,50,x)
        a1.append([ruslt1])
    for x in c:
        ruslt2=ConDegree(index_zz500,derr2,0,50,x)
        a2.append([ruslt2])
    for x in c:
        ruslt3=ConDegree(index_zz500,derr3,0,50,x)
        a3.append([ruslt3])
    for x in c:
        ruslt4=ConDegree(index_zz500,derr4,0,50,x)
        a4.append([ruslt4])
    for x in c:
        ruslt5=ConDegree(index_zz500,derr5,0,50,x)
        a5.append([ruslt5])
    b1=DataFrame(a1,columns=[derr1],index=c)
    b2=DataFrame(a2,columns=[derr2],index=c)
    b3=DataFrame(a3,columns=[derr3],index=c)
    b4=DataFrame(a4,columns=[derr4],index=c)
    b5=DataFrame(a5,columns=[derr5],index=c)
    b=pd.concat([b1,b2,b3,b4,b5],axis=1)
    return b
```
