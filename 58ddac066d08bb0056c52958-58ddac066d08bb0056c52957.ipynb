# 量价择时

# 基于华泰量价择时模型的长期择时实现
根据华泰行业周期分析，市场成交量与价格的同比序列结合对长期择时有很好的互补性，成交量同比序列和价格同比序列在同比序列大于 0 并且加速上升的时候对应的指数价格也处在一个强势的上升趋势中，因此把握这一部分确定的上涨机会也成为价量同比择时策略的基础。
下面根据该模型原理在通联平台进行分析。
首先是同比序列的定义。
价格同比序列为当月收盘价与一年前的收盘价的相对比值，即Pi/P(i-12),用对数表示即ln(Pi)-ln(P(i-12))
成交量取含当月的12个月的成交量之和，即Vi=∑k=i-11..i Vk
成交量同比序列为当月Vi与12月前Vi之比，即ln(Vi)-ln(V(i-12))
根据该模型定义，对每一行业计算当月同比序列，当满足价格同比序列和成交量同比序列相对上月均上升且成交量同比序列大于0（即Vi>Vi-12）的行业大于5个时说明牛市即将来临。
下面根据申万一级行业进行测试：

```python
import time
import datetime
def n_years_ago(dt,n):#取得前n年的日期
    year=dt[:4]
    ago_year=str(int(year)-n)
    ago_date=ago_year+dt[4:]
    return ago_date
```

```python
#参数：tradeDate:当前日期。index:指数。注意tradeDate格式：20100831
import math
import pandas as pd
def row_stastic(x):#计算单行业是否同时满足价格同比序列和成交量同比序列相对上月均上升且成交量同比序列大于0，是则返回1，否则返回0
    tmp=x['SPI']+x['SVI']+x['SVS']
    if tmp==3:
        return 1
    else:
        return 0
def S_VP(tradeDate,index,n):#计算单行业同比序列
    start_date=n_years_ago(tradeDate,n)
    end_date=tradeDate
    data=DataAPI.MktIdxmGet(beginDate=start_date,endDate=end_date,
                            ticker=index,field=u"endDate,ticker,closePrice,turnoverVol",
                            pandas="1")
    #计算价格序列
    data['logP']=data['closePrice'].apply(math.log)
    data['SP']=data['logP']-data['logP'].shift(-13)
    #计算成交量序列
    #首先计算过去12月成交量之和
    data=data.sort(['endDate'])
    data['V12']=pd.rolling_sum(data['turnoverVol'],12)
    data=data.sort(['endDate'],ascending=False)
    #计算成交量同比序列
    data['logV']=data['V12'].apply(math.log)
    data['SV']=data['logV']-data['logV'].shift(-13)
    #价格同比序列是否上升
    data['SPI']=(data['SP']-data['SP'].shift(-1))/abs(data['SP']-data['SP'].shift(-1))
    #价格同比序列是否大于0
    data['SPS']=data['SP']/abs(data['SP'])
    #成交量同比序列是否上升
    data['SVI']=(data['SV']-data['SV'].shift(-1))/abs(data['SV']-data['SV'].shift(-1))
    #成交量同比序列是否大于0
    data['SVS']=data['SV']/abs(data['SV'])
    data['result']=data.apply(row_stastic,axis=1)
    #data.to_csv(index+'.csv')
    #return data[['endDate','ticker','SP','SV','SPI','SPS','SVI','SVS','result']]
    return data[['endDate','result']]
```

先用5年的时间看看这个指标的反映。

```python
df_sw=pd.read_csv('申万一级.csv')
index_sw=df_sw['indexSymbol']

df=pd.DataFrame()
for index in index_sw:
    if len(df)==0:
        df=S_VP('20170321',str(index),5)
        #print(df)
    else:
        df_tmp=S_VP('20170321',str(index),5)
        #print(df_tmp)
        df=pd.merge(df,df_tmp,left_on='endDate',right_on='endDate',how='left')
df['result']=df.sum(axis=1,numeric_only=True)
```

```python
df=df.set_index('endDate')
```
