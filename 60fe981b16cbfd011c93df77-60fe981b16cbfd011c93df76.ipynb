# loc,iloc,ix的区别

```python
#loc 函数主要通过行标签  索引  行数据，重点：标签！标签！前闭后闭！
#iloc 主要是通过行号 获取 行数据，重点：序号！序号！序号！前闭后开！ 它只支持int型

import pandas as pd  
df = pd.DataFrame([  
            ['green', 'M', 10.1, 'class1'],   
            ['red', 'L', 13.5, 'class2'],   
            ['blue', 'XL', 15.3, 'class1']])
print(df)
print(df.loc[0])#0行

print('df.loc[0:1]:')
print(df.loc[0:1])#0行，1行，前闭加后闭

print('df.loc[0,1]:')
print(df.loc[0,1])

print('df.loc[:,0:2]:')
print(df.loc[:,0:2])

print('df.iloc[0:1]:')#iloc 主要是通过行号获取行数据，重点：序号！序号！序号！
print(df.iloc[0:1])
```

```python
import pandas as pd  
data = [[1,2,3],[4,5,6]]  
index = [0,1]  
columns=['a','b','c']  
df = pd.DataFrame(data=data, index=index, columns=columns)  
print(df)
print df.loc[1]

```
