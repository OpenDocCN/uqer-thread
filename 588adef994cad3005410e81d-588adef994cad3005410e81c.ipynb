- 卧槽，每次处理数据都有nan，这个玩意儿怎么处理的？下面有4个函数来处理这些未知数，应该能帮到你

```python
dt.ix[1:2,0:2] = np.nan#赋值nan
dt
```

```python
dt.dropna()#直接删除nan的行
```

```python
dt.fillna(0)#nan填充0
```

```python
dt.ffill()#延续上面的数据一直填充
```

这个ffill有啥用？如果你的index是时间序列，用ffill进行nan的填充会延用上一次数据，对于金融时间序列是一种便捷处理nan的函数

```python
dt.isnull()#判断nan的bool值
```

```python
dt.notnull()#和上面一个意思
```

总结下，上面可以找到标识nan的bool函以及nan的赋值和处理

- 之前说过的层次化索引能再讲讲么？我之前有说过？

```python
dt = pd.DataFrame({'4':[5,4,2],'8':[5,3,8],'5':[7,4,8]},index=[['one','one','three'],['four','five','six']])#造一个多索引dataframe
dt
```

```python
dt.index#多索引看上去还是有点不同，levels是索引值，labels是值的标识位
```

```python
dt.unstack()#这个有点像打包拆包
```

```python
dt.stack()#同上
```

```python
dt.index.names = ['1','2']
dt.columns.name = '3'
dt#可以赋予名字,这里name不行就换names……
```

```python
dt.swaplevel('1','2')#交换索引值位置
```

```python
dt.sum(level='1')#取索引并按列想加
```

```python
dt.set_index(['4','5'])#列转换为index
```

```python
dt.reset_index()#index转换为列
```
