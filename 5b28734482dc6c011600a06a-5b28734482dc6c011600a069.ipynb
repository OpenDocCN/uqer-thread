## 一、因子的灵感及构造逻辑

股市有一句俗语叫“量在价先”，意味着成交量有时候会先于价格出现波动，而这种现象在A股市场中并不少见，回头看常常会发现在利空/利好消息公布之前，对应股票的成交量已经发生了异动，这代表
知情交易者已经提前行动，例如：

- 贵州茅台在2017/10/26这一天的成交量异常抬升，达到前一日的4倍，当晚公布的三季报显示公司盈利超预期，提前完成年度目标，第二天股票大涨7.36%；
- 乐视网在2015/12/04这一天成交量骤增，当天换手率达到13.7%，次日公司宣布停牌进行资产重组，一停就是半年；
- 2013/09/10-2013/09/1，申华控股交易量激增，11日晚间公司宣布停牌并有重大资产重组举动，证监会确认其为内幕交易案件，相关责任人受到惩罚。
![图片注释](http://storage-uqer.datayes.com/5ad55b5af4ae710156a2bb13/3009f526-6ee3-11e8-8c58-0242ac140002)

基于A股“量在价先”的这类现象，我们可以做一个统计，统计根据换手率分组下不同组别涨跌幅绝对值的大小，结果发现**随着换手率均值的上升，涨跌幅的绝对值也呈上升趋势，这个结果表明市场的确
存在信息泄露的现象**，我们希望通过量价关系捕捉到价格异动的先兆，规避知情交易带来博弈上的劣势，因此我们构造了FR（Front Runnint）抢跑因子。

**因子逻辑**

我们在每月底回看过去一个月的交易数据，计算T日换手率与T+1日涨跌幅的关系，即FR = corr(turn, |ret|),其中turn是T日换手率，ret是T+1的涨跌幅的绝对值，corr是Pearson相关系数，在计算的
时候我们会剔除一月中停牌日数过多的股票；从表达式来看，因为FR是一个量价因子，为了验证FR是新的Alpha因子，我们在截面上对常见的因子(自由流通市值对数、20日动量、20日均换手率、60日波动
率)及行业因子(这里采用申万一级行业)进行回归剔除他们的影响，得到的残差即为纯净的FR因子，记为pure_fr。

## 二、因子收益分解优化

这里我们为了处理方便，我会同时计算FR与第二部分的收益分解优化后的因子。为了考察在A股大涨前和大跌前的信息泄露程度是否相同，信息泄露的因子在好消息和坏消息上是否具有相同的alpha，我们将
价格波动分解为上涨和下跌，我们在FR因子的逻辑基础上将过去一个月的交易数据根据**T+1日的**的涨跌幅分为两部分，如下图所示
![图片注释](http://storage-uqer.datayes.com/5ad55b5af4ae710156a2bb13/2a99fbd4-6f08-11e8-8c58-0242ac140002)
利用相对上涨的数据计算相关性，在横截面上回归掉市值、动量、换手率、波动率和行业，记为FRu,同理利用下跌的数据计算FRd，计算区间为(20080131-20180531)，月频更新，计算各自的原始IC及pure)IC,计算代码如下。

```python
import numpy as np
import pandas as pd
import datetime
import time
import os
import copy
import matplotlib.pyplot as plt
import statsmodels.api as sm
import scipy.stats as st
from CAL.PyCAL import *    # CAL.PyCAL中包含font
universe = set_universe('A')
cal = Calendar('China.SSE')
import seaborn as sns
sns.set_style('whitegrid')
```

```python
# 股票池文件存放目录，如果目录不存在，程序自动新建一个
raw_data_dir = "./FR_factor"
if not os.path.exists(raw_data_dir):
    os.mkdir(raw_data_dir)

def time_change(x):
    y = datetime.datetime.strptime(x, '%Y-%m-%d')
    y = y.strftime('%Y%m%d')
    return y

# 获取回测区间的交易日、月末以及月初时间
def get_trade_list(start_date, end_date):
    """
    Args:
        start_date: 时间区间起点
        end_date: 时间区间终点
    Returns: 
        trade_list: 时间区间内的交易日
        month_end: 月末时间
        month_start: 月初时间
    """
    cal_dates = DataAPI.TradeCalGet(exchangeCD=u"XSHG", beginDate=start_date, endDate=end_date).sort('calendarDate')
    cal_dates = cal_dates[cal_dates['isOpen'] == 1]
    trade_list = cal_dates['calendarDate'].values.tolist()
    trade_list = [time_change(x) for x in trade_list]
    month_end = cal_dates[cal_dates['isMonthEnd'] == 1]
    month_end = month_end['calendarDate'].values.tolist()
    month_end = [time_change(x) for x in month_end]
    cal = Calendar('China.SSE')
    month_start = [cal.advanceDate(x, '1B').strftime('%Y%m%d') for x in month_end]
    return trade_list, month_end, month_start

# 剔除ST股票
def st_remove(source_universe, st_date=None):
    """
    Args:
        source_universe (list of str): 需要进行筛选的股票列表
        st_date (datetime): 进行筛选的日期,默认为调用当天
    Returns:
        list: 去掉ST股票之后的股票列表
    """
    st_date = st_date if st_date is not None else datetime.datetime.now().strftime('%Y%m%d')
    df_ST = DataAPI.SecSTGet(secID=source_universe, beginDate=st_date, endDate=st_date, field=['secID'])
    return [s for s in source_universe if s not in list(df_ST['secID'])]

# 剔除某个日期前多少个交易日,之后上市的新股
def new_remove(ticker,tradeDate= None,day = 1):
    """
    Args:
        ticker (list of str): 需要进行筛选的股票列表（无后缀）
        tradeDate (datetime): 进行筛选的日期,默认为调用当天
        day (int): 向前漂移的交易日的个数
    Returns:
        list: 去掉新股股票之后的股票列表（无后缀）
    """
    tradeDate = tradeDate if tradeDate is not None else datetime.datetime.now()
    period = '-' + str(day) + 'B'
    pastDate = cal.advanceDate(tradeDate,period)
    pastDate = pastDate.strftime("%Y-%m-%d")
    ipo_date = DataAPI.SecIDGet(partyID=u"",assetClass=u"e",ticker=ticker,cnSpell=u"",field=u"ticker,listDate",pandas="1")
    remove_list = ipo_date[ipo_date['listDate'] > pastDate]['ticker'].tolist()
    return [stk for stk in ticker if stk not in remove_list]

# 将股票代码转化为股票内部编码
def ticker2secID(ticker):
    """
    Args:
        tickers (list): 需要转化的股票代码列表
    Returns:
        list: 转化为内部编码的股票编码列表
    """
    universe = DataAPI.EquGet(equTypeCD=u"A",listStatusCD="L,S,DE,UN",field=u"ticker,secID",pandas="1") # 获取所有的A股（包括已退市）
    universe = dict(universe.set_index('ticker')['secID'])
    if isinstance(ticker, list):
        res = []
        for i in ticker:
            if i in universe:
                res.append(universe[i])
            else:
                print i, ' 在universe中不存在，没有找到对应的secID！'
        return res
    else:
        raise ValueError('ticker should be list！')

# 获取股票池函数
def get_stock_pool(date, N):
    """
    Args:
        date: 月初时间
        N: 新股的定义时间
    Returns:
        stock_pool: 此月初的股票池
    """
    univ=DynamicUniverse('A')
    all_code = univ.preview(date,skip_halted=False)
    all_code_not_ST = st_remove(all_code, st_date=date)
    ticker = [x.split('.')[0] for x in all_code_not_ST]
    all_code_need = new_remove(ticker, tradeDate=date, day=N)
    code = ticker2secID(all_code_need)
    df = pd.DataFrame({'code': code})
    df['date'] = date
    df = df[['date', 'code']]
    return df

# 获取每月初的股票池(20080102 - 20180601)
start_date = '20071220'
end_date = '20180601'        
trade_list, month_end, month_start = get_trade_list(start_date, end_date)
N = 180
all_stock = pd.DataFrame()
for date in month_start:
    stock = get_stock_pool(date, N)
    all_stock = pd.concat([all_stock, stock],axis=0)

all_stock.to_csv('FR_factor/stock_pool.csv', index=False)
print('***********股票池示例************')
print(all_stock.head(10).to_html())
```

```python
# 行情数据
all_code = sorted(list(set(all_stock['code'])))
mkt_info_frame = DataAPI.MktEqudGet(secID=all_code,beginDate=u"20071120",endDate=u"20180601",isOpen="1",field=u"secID,tradeDate,chgPct,turnoverRate",pandas="1")
mkt_info_frame['tradeDate'] = map(time_change, mkt_info_frame['tradeDate'])
mkt_info_frame.columns = ['code', 'date', 'ret', 'to']
mkt_info_frame = mkt_info_frame.dropna()
```

```python
# 构造因子FR、FRu、FRd
all_signal = pd.DataFrame()
for i in range(1, len(month_end)):
    start = month_start[i]
    end = month_end[i]
    begin = month_start[i - 1]
    # 次月收益
    if i < len(month_end) - 1:
        next_end = month_end[i + 1]
    else:
        next_end = end
    code = list(all_stock[all_stock['date'] == start]['code'])
    close1 = DataAPI.MktEqudAdjAfGet(secID=code,tradeDate=end,field=u"secID,closePrice",pandas="1")
    close1.columns = ['code', 'close1']
    close2 = DataAPI.MktEqudAdjAfGet(secID=code,tradeDate=next_end,field=u"secID,closePrice",pandas="1")
    close2.columns = ['code', 'close2']
    close = pd.merge(close1, close2, on='code', how='outer')
    close['month_ret'] = close['close2'] / close['close1'] - 1
    close['date'] = end
    close = close[['date', 'code', 'month_ret']]
    # 因子
    period_frame = mkt_info_frame[(mkt_info_frame['date'] >= begin) & (mkt_info_frame['date'] <= end)]
    period_frame = period_frame[period_frame['code'].isin(list(all_stock[all_stock['date'] == start]['code']))]
    period_frame['ret_next'] = period_frame.groupby('code')['ret'].shift(-1)
    period_frame['abs_ret_next'] = abs(period_frame['ret_next'])   
    period_frame.dropna(inplace=True)
    trading_day = [x for x in trade_list if x >= begin and x < end]
    n_days = len(trading_day) / 2
    # FR
    signal_fr = period_frame.groupby('code').apply(lambda df: df['to'].corr(df['abs_ret_next'], method='pearson', min_periods=n_days))
    # FRu
    period_frame_positive = period_frame[period_frame['ret_next'] > 0]
    signal_fru = period_frame_positive.groupby('code').apply(lambda df: df['to'].corr(df['ret_next'], method='pearson', min_periods=5))
    # FRd
    period_frame_negative = period_frame[period_frame['ret_next'] <= 0]
    signal_frd = period_frame_negative.groupby('code').apply(lambda df: df['to'].corr(df['ret_next'], method='pearson', min_periods=5))
    # 整合
    signal = pd.concat([signal_fr, signal_fru], axis=1, join='outer')
    signal = pd.concat([signal, signal_frd], axis=1, join='outer')
    signal.columns = ['fr', 'fru', 'frd']
    signal['code'] = signal.index
    signal['date'] = end
    signal = signal[['date', 'code', 'fr', 'fru', 'frd']]
    signal.sort_values(by='code', inplace=True)
    signal.reset_index(drop=True, inplace=True)
    signal = pd.merge(close, signal, how='right')
    all_signal = all_signal.append(signal) # all_signal就是计算出的三个原始因子FR,FRu,FRd
```

```python
# 测试因子IC
def cal_ic(factor_name, data):
    IC = []    
    Month_end = sorted(list(set(data['date'])))
    for date in Month_end:
        tdata = data[data['date'] == date]
        tdata.sort_values(by='code', inplace=True)
        tdata.reset_index(drop=True, inplace=True)        
        factor = np.array(tdata[factor_name])
        ret = np.array(tdata['month_ret'])
        cor, pval = st.spearmanr(factor, ret)
        IC.append(cor)
    IC_data = pd.DataFrame({'date': Month_end, 'IC': IC})
    IC_data = IC_data[['date', 'IC']]
    return IC_data

raw_IC_fr = cal_ic('fr', all_signal)
raw_IC_fr.columns = ['date', 'IC_fr']
raw_IC_fru = cal_ic('fru', all_signal)
raw_IC_fru.columns = ['date', 'IC_fru']
raw_IC_frd = cal_ic('frd', all_signal)
raw_IC_frd.columns = ['date', 'IC_frd']
raw_IC = pd.merge(raw_IC_fr, raw_IC_fru, on='date')
raw_IC = pd.merge(raw_IC, raw_IC_frd, on='date')
raw_IC = raw_IC[8: -1]
print 'FR因子的IC，IR分别是: %f，%f' % (raw_IC['IC_fr'].mean(), raw_IC['IC_fr'].mean() / raw_IC['IC_fr'].std() * np.sqrt(12))
print 'FRu因子的IC，IR分别是: %f，%f' % (raw_IC['IC_fru'].mean(), raw_IC['IC_fru'].mean() / raw_IC['IC_fru'].std() * np.sqrt(12))
print 'FRd因子的IC，IR分别是: %f，%f' % (raw_IC['IC_frd'].mean(), raw_IC['IC_frd'].mean() / raw_IC['IC_frd'].std() * np.sqrt(12))
```

```python
# 获取某个时点股票所属行业
def get_industry(date, code):
    """
    Args:
        date: 月初时间
        code: 股票代码列表，如['000001.XSHE', '000002.XSHE']
    Returns:
        indu: 申万一级行业因子数据,dataframe格式，列名为日期，股票代码，行业名
    """ 
    cal = Calendar('China.SSE')
    end = cal.advanceDate(date, '-1B').strftime('%Y%m%d') 
    indu = DataAPI.MdSwBackGet(secID=code,field=u"secID,isNew,industryName1",pandas="1")
    indu = indu[indu['isNew'] == 1]
    indu.drop_duplicates(subset=['secID'], inplace=True)
    indu['date'] = date
    indu = indu[['date', 'secID', 'industryName1']]
    indu.columns = ['date', 'code', 'industry']
    return indu

# 计算纯净因子
all_pure_fr = pd.DataFrame()
for date in month_end[1: ]:
    tdf = all_signal[all_signal['date'] == date]
    code = list(tdf['code'])
    # 优矿因子库的因子：对数自由流通市值、20日动量、20日换手率、60日波动率
    factors = DataAPI.MktStockFactorsOneDayProGet(tradeDate=date,secID=code,field=u"secID,tradeDate,lflo,revs20,vol20,variance60",pandas="1")
    factors['tradeDate'] = map(time_change, factors['tradeDate'])
    factors.columns = ['code', 'date', 'lflo', 'revs20', 'to20', 'vol60']
    tdf = pd.merge(tdf, factors, on=['date', 'code'], how='left')
    # 行业，这里选用申万一级行业
    indu = get_industry(date, code)
    tdf = pd.merge(tdf, indu, on=['date', 'code'], how='left')
    tdf = tdf.dropna()
    # 去极值 + 标准化
    tdf = tdf.set_index('code')
    for col in tdf.columns[2: -1]:
        temp = tdf[col]
        tdf[col] = standardize(winsorize(temp))
    # 将行业转换成虚拟变量
    indu_dummies = pd.get_dummies(tdf['industry'])
    del tdf['industry']
    tdf = pd.concat([tdf, indu_dummies], axis=1)
    # 回归
    y1 = np.array(tdf['fr'])
    y2 = np.array(tdf['fru'])
    y3 = np.array(tdf['frd'])
    x = np.array(tdf[tdf.columns[5: ]])
    x = sm.add_constant(x, has_constant='add')
    # pure_fr
    model = sm.OLS(y1, x, missing='drop')
    results = model.fit()
    tdf['fr'] = results.resid
    tdf['fr'] = standardize(winsorize(tdf['fr']))
    # pure_fru
    model = sm.OLS(y2, x, missing='drop')
    results = model.fit()
    tdf['fru'] = results.resid
    tdf['fru'] = standardize(winsorize(tdf['fru']))    
    # pure_frd
    model = sm.OLS(y3, x, missing='drop')
    results = model.fit()
    tdf['frd'] = results.resid
    tdf['frd'] = standardize(winsorize(tdf['frd']))
    
    tdf['code'] = tdf.index
    pure_fr = tdf[['date', 'code', 'fr', 'fru', 'frd', 'month_ret']]
    pure_fr.reset_index(drop=True, inplace=True)
    all_pure_fr = all_pure_fr.append(pure_fr)
    
all_pure_fr.to_csv('FR_factor/pure_fr.csv', index=False) # 纯净因子存储
```

```python
pure_IC_fr = cal_ic('fr', all_pure_fr)
pure_IC_fr.columns = ['date', 'IC_fr']
pure_IC_fru = cal_ic('fru', all_pure_fr)
pure_IC_fru.columns = ['date', 'IC_fru']
pure_IC_frd = cal_ic('frd', all_pure_fr)
pure_IC_frd.columns = ['date', 'IC_frd']
pure_IC = pd.merge(pure_IC_fr, pure_IC_fru, on='date')
pure_IC = pd.merge(pure_IC, pure_IC_frd, on='date')
pure_IC_cal = pure_IC.copy()
pure_IC_cal = pure_IC[: -1]
pure_IC = pure_IC[8: -1]
pure_IC.reset_index(drop=True, inplace=True)
print 'pure_FR因子的IC，IR分别是: %f，%f' % (pure_IC['IC_fr'].mean(), pure_IC['IC_fr'].mean() / pure_IC['IC_fr'].std() * np.sqrt(12))
print 'pure_FRu因子的IC，IR分别是: %f，%f' % (pure_IC['IC_fru'].mean(), pure_IC['IC_fru'].mean() / pure_IC['IC_fru'].std() * np.sqrt(12))
print 'pure_FRd因子的IC，IR分别是: %f，%f' % (pure_IC['IC_frd'].mean(), pure_IC['IC_frd'].mean() / pure_IC['IC_frd'].std() * np.sqrt(12))
```

从原始IC以及pureIC的角度来看，我们可以得到如下结论：

- FR是一个反向因子，意味着信息泄漏最低的组后续收益最高，这与我们的猜想一致；
- 对比IC发现，FRd > FR > FRu,这说明知情交易者在好消息和坏消息上的泄露程度不同，坏消息的泄露现象更为显著；
- FRu因子的多头来自于第一组负相关组，空头来自于第五组正相关组；FRd的多头来自于第五组正相关组，空头来自于第一组负相关组；
- 通过收益分解来对FR因子进行增强，我们可以用FRd代替FR因子，或者考虑到FRu和FRd的关系，对两者进行加权合成增强后的复合因子。

```python
def run_backtest(factor_name, backtest_data, signal_df, neutralize=True):
    t1 = time.time()
    
    factor_data = signal_df[['secID', 'tradeDate', factor_name]]     # 读取因子数据
    # factor_data[factor_name] = -factor_data[factor_name]
    factor_data = factor_data.set_index('tradeDate', drop=True)
    month_end_dates = factor_data.index.values
    t2 = time.time()
    print "initial time cost",t2-t1
    # 运行结果
    results = {}

    # # 将因子划分为5分位，并进行快速回测
    for quantile_five in range(1, 6):
        # ---------------策略逻辑部分----------------
        # 初始化虚拟账户状态
        def initialize(context):
            pass
    
        # 每个调仓期执行一次
        def handle_data(context): 
            account = context.get_account('security_account')
            pre_date = context.previous_date.strftime("%Y%m%d")
            current_universe = context.get_universe(exclude_halt=True)
            current_universe = st_remove(current_universe, st_date=pre_date)
            current_universe = new_remove(current_universe, tradeDate= pre_date)
            if pre_date not in month_end_dates: 
                return

            # 拿取调仓日前一个交易日的因子，并按照大小排列取相应分位的股票代码
            q = factor_data.ix[pre_date]
            q = q.set_index('secID', drop=True)
            q = q.ix[current_universe]
            q.dropna(inplace=True)
            
            if neutralize:
                # he shizhizuo zhengjiao
                signal = q[factor_name]
                style_indu = DataAPI.RMExposureDayGet(tradeDate=pre_date.replace("-","")).set_index('secID')
                style_indu = style_indu.ix[signal.index, 6:7]
                ids = signal.index
                y = signal.ix[ids].values
                X = np.column_stack([style_indu.loc[signal.index,:].values])
                X = sm.add_constant(X)

                mod_wls = sm.OLS(y, X)
                mod_fit = mod_wls.fit()
                res_params = mod_fit.params
                res_params[pd.isnull(res_params)] = 0
                residual_vec = y - np.dot(X, res_params)
                q = pd.Series(residual_vec, index=ids)
                q = pd.DataFrame(data=q.values,index=q.index.values, columns=[factor_name])
            
            q_min = q[factor_name].quantile((quantile_five-1)*0.2)
            q_max = q[factor_name].quantile(quantile_five*0.2)
            my_univ = q[(q[factor_name] >= q_min) & (q[factor_name] < q_max)].index.values

           # 将不在目标持仓中的股票卖出
            positions = account.get_positions()
            sell_list = [stk for stk in positions if stk not in my_univ]
            for stk in sell_list:
                account.order_to(stk,0)

            # 将在目标股票池中的股票，等权买入
            for stk in my_univ:
                account.order_pct_to(stk, 1.0/len(my_univ))

        # 生成策略对象
        strategy = quartz.TradingStrategy(initialize, handle_data)

        # 开始回测
        bt, perf = quartz.quick_backtest(sim_params, strategy, data=backtest_data)

        # 保存运行结果，1为因子最强组，5为因子最弱组
        results[6 - quantile_five] = {'max_drawdown': perf['max_drawdown'], 
                                      'sharpe': perf['sharpe'], 
                                      'alpha': perf['alpha'], 
                                      'beta': perf['beta'], 
                                      'information_ratio': perf['information_ratio'], 
                                      'annualized_return': perf['annualized_return'], 
                                      'bt': bt}    

        print str(quantile_five),

    print 'done'
    print "Time cost: %s seconds" %(time.time() - t1)
    return results

def draw_graphics(results):    
    sns.set_style('white')

    fig = plt.figure(figsize=(10,8))
    fig.set_tight_layout(True)
    ax1 = fig.add_subplot(211)
    # ax2 = fig.add_subplot(212)
    ax1.grid()
    # ax2.grid()

    temp = results[1]['bt']
    temp = temp[['tradeDate']]
    
    for qt in results:
        bt = results[qt]['bt']

        data = bt[[u'tradeDate',u'portfolio_value',u'benchmark_return']]
        data['portfolio_return'] = data.portfolio_value/data.portfolio_value.shift(1) - 1.0   # 总头寸每日回报率
        data['portfolio_return'].ix[0] = data['portfolio_value'].ix[0]/	100000000.0 - 1.0
        data['excess_return'] = data.portfolio_return - data.benchmark_return                 # 总头寸每日超额回报率
        data['excess'] = data.excess_return + 1.0
        data['excess'] = data.excess.cumprod()                # 总头寸对冲指数后的净值序列
        data['portfolio'] = data.portfolio_return + 1.0     
        data['portfolio'] = data.portfolio.cumprod()          # 总头寸不对冲时的净值序列
        data['benchmark'] = data.benchmark_return + 1.0
        data['benchmark'] = data.benchmark.cumprod()          # benchmark的净值序列
        results[qt]['hedged_max_drawdown'] = max([1 - v/max(1, max(data['excess'][:i+1])) for i,v in enumerate(data['excess'])])  # 对冲后净值最大回撤
        results[qt]['hedged_volatility'] = np.std(data['excess_return'])*np.sqrt(252)
        results[qt]['hedged_annualized_return'] = (data['excess'].values[-1])**(252.0/len(data['excess'])) - 1.0
        ax1.plot(data['tradeDate'], data[['portfolio']], label=str(qt))
        # ax2.plot(data['tradeDate'], data[['excess']], label=str(qt))
        ret = data[['tradeDate', 'portfolio_return']]
        ret.columns = ['tradeDate', 'ret_' + str(qt)]
        temp = pd.merge(temp, ret, on='tradeDate')

    # 1-5多空对冲
    temp['hedge_1_to_5'] = (temp['ret_1'] - temp['ret_5'])
    temp['portfolio'] = (1 + temp['hedge_1_to_5']).cumprod()   
    temp.to_csv('FR_factor/1_5_hedge_FRd.csv', index=False)
        
    ax1.legend(loc=0)
    # ax2.legend(loc=0)
    ax1.set_ylabel(u"净值", fontproperties=font, fontsize=16)
    # ax2.set_ylabel(u"对冲净值", fontproperties=font, fontsize=16)
    ax1.set_title(u"因子不同五分位数分组选股净值走势", fontproperties=font, fontsize=16)
    # ax2.set_title(u"因子不同五分位数分组选股对冲中证500指数后净值走势", fontproperties=font, fontsize=16)

    # results 转换为 DataFrame
    results_pd = pd.DataFrame(results).T.sort_index()

    results_pd = results_pd[[u'alpha', u'beta', u'information_ratio', u'sharpe', u'annualized_return', u'max_drawdown',  
                             u'hedged_annualized_return', u'hedged_max_drawdown', u'hedged_volatility']]

    cols = [(u'风险指标', u'Alpha'), (u'风险指标', u'Beta'), (u'风险指标', u'信息比率'), (u'风险指标', u'夏普比率'), (u'纯股票多头时', u'年化收益'),
            (u'纯股票多头时', u'最大回撤'), (u'对冲后', u'年化收益'), (u'对冲后', u'最大回撤'), (u'对冲后', u'收益波动率')]
    results_pd.columns = pd.MultiIndex.from_tuples(cols)
    results_pd.index.name = u'五分位组别'
    return results_pd
```

```python
def plot_under_water_enhance(data, title):
    """
    绘制回撤及收益率曲线图
    输入：
        bt：quartz回测结束自动生成的dict
        title：str
    返回：
        ax：matplotlib figure 对象
    """
    df_cum_rets = data['hedge_1_to_5']
    running_max = np.maximum.accumulate(df_cum_rets)
    underwater = -((running_max - df_cum_rets) / running_max)
    underwater.index = pd.to_datetime(data['tradeDate'])

    fig = plt.figure(figsize=(12, 5))
    fig.set_tight_layout(True)
    ax1 = fig.add_subplot(111)
    ax2 = ax1.twinx()
    x = range(len(underwater))
    ax2.grid(False)
    ax1.set_ylim(-0.30, 0)
    ax1.set_ylabel(u'回撤', fontproperties=font, fontsize=16)
    ax1.fill_between(underwater.index, 0, np.array(underwater), color='#000066', alpha=1)
    ax2.set_ylabel(u'净值', fontproperties=font, fontsize=16)
    ax2.plot(data['tradeDate'], data[['hedge_1_to_5']], label='hedged(right)', color='r')
    ax2.set_ylim(bottom=0.9, top=3)
    s = ax1.set_title(title, fontproperties=font, fontsize=16)
    ax1.grid(True)
    return fig

ret = pd.read_csv('FR_factor/1_5_hedge_FRd.csv')
ret = ret[['tradeDate', 'hedge_1_to_5']]
ret.columns = ['tradeDate', 'hedge_1_to_5']
ret['tradeDate'] = pd.to_datetime(ret['tradeDate'])
ret['hedge_1_to_5'] = (ret['hedge_1_to_5'] + 1).cumprod()
plot_under_water_enhance(ret, title=u"多空对冲净值曲线")
```

这是FRd因子的多空对冲净值曲线，曲线很稳健，而由于pure_FRd剥离掉了市值、反转、波动率、换手率、行业等因子的影响，是除上述因素之外的Alpha源，因此该因子的多空收益不是很高。下面我们
尝试结合pure_FRu与pure_FRd采用最大化IC_IR的方法对因子进行增强。（观察窗口取12期，协方差矩阵用Ledoit_Wolf法进行压缩）

```python
# N天IC_IR方法合成的复合因子
from sklearn.covariance import LedoitWolf
import numpy.linalg as nlg
# 加权优化
def get_period_data(currentdate, df, N): # N代表滚动窗口长度

    l_date = sorted(list(set(df['date'])))
    date = [l_date[l_date.index(currentdate) - i] for i in range(N)]
    wdata = df[df['date'].isin(date)]
    return wdata

date_list = sorted(list(set(pure_IC_cal['date'])))
df = pd.DataFrame()
del pure_IC_cal['IC_fr']
N = 12 # 建立模型的时间跨度
for i in range(N, len(date_list)):
    date = date_list[i - 1]
    currentdate = date_list[i] # 当期日期
    # N期IC数据
    wdata = get_period_data(date, pure_IC_cal, N)
    period_date = list(wdata['date'])
    wdata.set_index('date', inplace=True)
    mat = np.mat(wdata.cov())  # 样本协方差矩    
    LW = LedoitWolf().fit(wdata).covariance_ # 压缩协方差矩阵
    mat = nlg.inv(LW) 
    # 权重
    weight = mat * np.mat(wdata.mean()).reshape(len(mat), 1)
    weight= np.array(weight.reshape(len(weight),))[0]
    # 暴露
    tdata = all_pure_fr[all_pure_fr['date'] == currentdate]
    # expo = np.array(tdata.iloc[: , 3: 5])
    expo = np.array(tdata[['fru', 'frd']])
    weight = np.dot(expo, weight)
    tdata['IC_IR_FR'] = weight
    df = df.append(tdata)
    
mix_ic = cal_ic('IC_IR_FR', df)
mix_ic.columns = ['date', 'mix_IC']
print 'IC_IR加权复合FR因子的IC，IR分别是: %f，%f' % (mix_ic['mix_IC'].mean(), mix_ic['mix_IC'].mean() / mix_ic['mix_IC'].std() * np.sqrt(12))
# IC序列图
fig = plt.figure(figsize=(16, 5))
ax = fig.add_subplot(111)
ax.bar(range(len(mix_ic)), mix_ic['mix_IC'], width=0.5, align='edge')
ax.set_xlim(0, len(mix_ic))
x_ticks = range(0, len(mix_ic), 12)
x_label = [mix_ic['date'][i] for i in x_ticks]
ax.set_xticks(x_ticks)
ls = ax.set_xticklabels(x_label, rotation=40)
ax.set_xlabel(u'日期', fontproperties=font, fontsize=15)
ax.set_ylabel(u'IC', fontproperties=font, fontsize=15)
ax.grid(True)
```
