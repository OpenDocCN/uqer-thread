```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
from CAL.PyCAL import *

# print nextDate
StkDay='20160831'    #股票交易集的日期
```

```python
#获取所有股票之后再获取所在行业
IndSTKbeforeNumeric = DataAPI.EquIndustryGet(industryVersionCD=u"010303",secID=DataAPI.EquGet(secID=set_universe('A'),listStatusCD=u"L",field=u"",pandas="1")['secID'].tolist(),intoDate=StkDay,field=['secID','industryName1'],pandas="1")

#改列名
IndSTKbeforeNumeric.columns = ['secID','Ind']
IndSTK = IndSTKbeforeNumeric

#提取行业列表
IndSTKbeforeNumeric2 = IndSTKbeforeNumeric.groupby(IndSTKbeforeNumeric['Ind'])
INDSTList = IndSTKbeforeNumeric2.any().index.values.tolist()

#提取dummies值
IndSTKdummies = pd.concat([IndSTK,pd.get_dummies(IndSTK['Ind'])],axis = 1)
```

```python
def FactorGLSttest(T_day,dayGap,FaList):
    #此函数 输入的是因子列表，输出包含列表中所有因子时的GLS结果，以DataFrame格式return
    Gap = dayGap
    tday = T_day

    cal = Calendar('China.SSE')
    Tday = Date.strptime(tday,dateFormat = '%Y%m%d')
    period = Period('-'+str(Gap)+'b')

    nextDate = cal.advanceDate(Tday, period, BizDayConvention.Following)

    previousDay = nextDate.strftime(strFormat = '%Y%m%d')


    FactorBetaPvalueTable = pd.DataFrame()
    IndSTKdummiesfac = IndSTKdummies
    

    for factName in FaList:
        fac = DataAPI.MktStockFactorsOneDayGet(tradeDate=previousDay,secID=IndSTKdummies['secID'],field=["secID"]+[factName],pandas="1")

        IndSTKdummiesfac = pd.merge(IndSTKdummiesfac,fac)

        ReveseList=['PE', 'PS', 'PB', 'PCF', 'PEG3Y']

        if factName in ReveseList:

            IndSTKdummiesfac.insert(0,'Stdardized_'+factName,((1.0/IndSTKdummiesfac[factName])-pd.np.mean((1.0/IndSTKdummiesfac[factName])))/pd.np.std((1.0/IndSTKdummiesfac[factName])))
            print 'Factor ' + factName +' Revesed!'

        else:
            IndSTKdummiesfac.insert(0,'Stdardized_'+factName,((IndSTKdummiesfac[factName])-pd.np.mean((IndSTKdummiesfac[factName])))/pd.np.std((IndSTKdummiesfac[factName])))
        
        del IndSTKdummiesfac[factName]

        
    # 取两个时间的closePrice，算pct
    
    ClosePrices = DataAPI.MktEqudAdjGet(tradeDate=Tday,secID=IndSTK['secID'],field=u"secID,closePrice",pandas="1")
    preClosePrices = DataAPI.MktEqudAdjGet(tradeDate=previousDay,secID=IndSTK['secID'],field=u"secID,closePrice",pandas="1")
    pct_t1 = ClosePrices['closePrice']/preClosePrices['closePrice']-1
    
    #
    
    ClosePrices.insert(0,'pct_t1',pct_t1)
    pctBpDummies = pd.merge(ClosePrices,IndSTKdummiesfac)#IndSTKDummiesFactor.insert(0,'pct_t1',pct_t1)

    pctBpDummies

    del pctBpDummies['closePrice']
    del pctBpDummies['preClosePrice']

    del pctBpDummies['Ind']
    del pctBpDummies['secID']

    del pctBpDummies['交通运输'] # 去掉一个行业


    pctBpDummies.insert(len(FaList)+1,'ones',np.ones(pctBpDummies.shape[0])) # 增加一列1

    pctBpDummies = pctBpDummies.dropna()
    
    # print pctBpDummies.head()
    
    gls_model = sm.GLS(pctBpDummies.iloc[:,0],pctBpDummies.iloc[:,1:]).fit()
    # print factName
    for factName in FaList:
        # print factName
        FactorBetaPvalueTable = pd.concat([FactorBetaPvalueTable,pd.DataFrame([factName,gls_model.pvalues['Stdardized_'+factName],gls_model.params['Stdardized_'+factName],gls_model.aic,gls_model.bic])],axis = 1) 
    
    FactorBetaPvalueTable = FactorBetaPvalueTable.T
    
    FactorBetaPvalueTable.index = range(FactorBetaPvalueTable.shape[0])
    FactorBetaPvalueTable.columns = ['FactorName','PValue','Beta','AIC','BIC']
    print FactorBetaPvalueTable.to_html()
    return FactorBetaPvalueTable
```

```python
list_P_B = ['LCAP','ILLIQUIDITY','MoneyFlow20','PB','PE','REVS20','VOL20','PS','PCF','CTOP']

# 去掉 Pvalue > 0.05的因子
for facc in list_P_B:
    
    B_P_Table = FactorGLSttest('20170421',1,[facc])
    if B_P_Table['PValue'].values >0.05:
        list_P_B.remove(facc)

BPTable2 = pd.DataFrame()        

# 剩下的因子每个都单独算一次，拼起来。
for facc in list_P_B:
    
    B_P_Table = FactorGLSttest('20170421',1,[facc])
    BPTable2 = pd.concat([BPTable2,B_P_Table])
    
```
