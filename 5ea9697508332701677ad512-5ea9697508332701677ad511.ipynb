濮老师认为，1，虽然该模型只能谈得上战胜指数，但是以如此简单的逻辑获得显然领先指数的绩效，我们应该意识到A股的反转特征已经是非常强烈了；2，二八止损模块的加入不仅是在股灾期间抗击回撤，在股灾之前，它也让系统最高达到了400%的收益率，并在2010~2014年逐步和基准指数拉开差距。

模型在2010-2015年表现非常好，但2018年以后非常差。所以，量化模型的及时调整，对业绩非常关键。
