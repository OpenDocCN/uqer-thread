```python
info_all = pd.read_csv('industry_matrix.csv')     #industry_matrix为行业矩阵，取值为0-1，可以自己取数据合成，同时剔除ST股，新股等
info_all.iloc[:,:15].tail(5)
```

```python
info_all['date'] = info_all['date'].apply(lambda x:str(x))
info_all['code'] = info_all['code'].apply(lambda x:str(x).zfill(6))
date = info_all['date'].drop_duplicates().values

result = pd.DataFrame()
for d in date:
    try:
        tickers = info_all[info_all['date'] == d]['code'].values
        r_tmp = Predict(tickers, d)  #预测
        result = result.append(r_tmp)
    except:
        print d
result.to_csv('alpha.csv',index=None)
```
