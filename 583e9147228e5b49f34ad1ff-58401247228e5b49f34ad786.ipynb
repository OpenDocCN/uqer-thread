### 下面把上次这个模型继续优化一下。

   > 机器学习有句话，叫做：数据和特征决定了机器学习的上限，而模型和算法只是逼近这个上限而已

比如说，要知道一个人长什么样，必须知道他脸的长宽，眼镜的位置，嘴巴的大小等等。但是如果给你一个人的学习成绩，血型等特征，那无论如何，都无法做出很好的模型来预测长相。

在量化学习中，这里的特征就是因子，必须找到和价格涨跌关联度最大的因子，做出的模型才能有效。

这个特征的选择每个人都有每个人的方法，我先介绍一些机器学习的基本技巧。

机器学习模型的训练，最好先对特征值进行`无量纲化`处理。

简单解释下无量纲化：比如特征值A的范围是0-10000的，特征值B的范围是0-100，就不如把这两组特征进行标准化，都转化成0-1的范围。

这种特征预处理，在sklearn中有专门的用来预处理数据的库。

* `from sklearn import preprocessing`

preprocessing有几种无量纲化的方法，常见的有`StandardScaler`,`MinMaxScaler`,`Normalizer`等等.

预处理也有fit过程，fit之后不是predict，而是transform

比如现在有训练数据train，测试数据test。

首先利用train数据fit和转化

* `preprocessing.Normalizer().fit_transform(train)`

* `preprocessing.Normalizer().transform(test)`

fit_transform是把fit和transform合并了。第二部是按照train数据的fit方法将test数据进行transform

下面将上面的策略里面的特征进行无量纲化，看看对机器学习的训练的帮助。

另外我还进行了去噪声处理（仅保留涨跌排序前后30%的数据，去除中间的噪声），然后二值化处理（涨幅前30%设置目标1，涨跌后30%设置目标0）。

这个结果比之前没有做任何预处理要优秀的多，至少拥有了正的alpha，而且模型的参数没有任何变化，说明数据和特征对优化模型预测效果有非常明显的效果。

本文是想告诉正在学习机器学习的朋友，多花一些心思在特征的选择上，就如上文所说`数据和特征决定了机器学习的上限，而模型和算法只是逼近这个上限而已`

对特征值进行无量纲化可以让机器学习的训练更加有效，下一篇我将继续探索降维的作用。

### 接下来说一下降维的作用


在上面的例子中，我们选取了

'LCAP', 'LFLO', 'NPToTOR', 'OperatingProfitGrowRate', 'TotalAssetGrowRate', 'DHILO', 'DEGM', 'Skewness', 'DAREC', 'GREC'

这10个因子，但是并不是每一个因子都是独立的。

比如LCAP和LFLO，分别是市值和流通市值的对数值，根据经验，这两个特征是高度正相关的，DAREC和GREC分别是分析师推荐评级变化和分析师推荐评级变化趋势，这两个特征也是拥有一定关联性的

因此在模型的训练中过多的特征会导致模型过于复杂，过拟合的风险也会加大，因此这个时候，我们需要对原特征进行降维。

sklearn中降维的方法主要有PCA,LDA。分别在以下位置

`from sklearn.decompostion import PCA`

`from sklearn.lda import LDA`

fit_transform方法和无量纲化类似。其中LDA需要target进行有监督的降维（LDA.fit_transform(train,target)）

下面我们对上面的策略进行PCA降维处理，降维后保留6个特征。

对数据进行PCA降维处理后发现，模型的有效性进一步的提高了。

增加了交易手续费之后，alpha依然提高到了13.3%，进一步证明了特征工程对于机器学习的重要性

至此为止，我把最初的一个粗糙的随机森林模型进行了如下几步优化：

* 1、数据的清洗，把最初一股脑儿放入训练的数据进行去噪声，只涨幅榜前30%和后30%的股票数据，去除中间的噪声。

* 2、对因子特征进行无量纲化。使得机器学习模型的训练更加容易。

* 3、对特征进行主成分分析，降维。防止模型过于复杂和过拟合。
