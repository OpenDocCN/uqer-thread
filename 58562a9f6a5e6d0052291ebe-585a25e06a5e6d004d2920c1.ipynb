```python
# 分类训练数据
def SNW_Quotes_Amplitude_Classification(universe=['000001.XSHE','600000.XSHG'], used_factors= ['PE','ADTM'],
                                        start_date='20160101', end_date='20161130', forecast_period=10):
    #转换时间格式
    start_date = parse(start_date)
    end_date = parse(end_date)
    
    #交易所交易日历
    cal = Calendar('China.SSE')
    trading_date_list = cal.bizDatesList(start_date,end_date)
    
    #取得股票日K线报价单，处理预测时间涨跌预测训练原始数据
    #这里使用前复权数据，个人感觉后复权貌似更有趣一点
    quotes_tmp= DataAPI.MktEqudGet(secID=universe, beginDate= start_date, endDate= end_date, isOpen="",field=['tradeDate','secID','closePrice'] , pandas="1")
    secID_list= sorted(list(set(quotes_tmp.secID.tolist())))
    cache_quotes= pd.DataFrame()
    for i in secID_list:
        t= quotes_tmp[quotes_tmp['secID']==i]
        tmp= copy.deepcopy(t)
        tmp['amplitude']= quotes_tmp['closePrice'].shift(-forecast_period)/quotes_tmp['closePrice'] - 1.0
        #此处会产生不同股票计算振幅的bug,修改 2016-12-20
        tmp = tmp[:-forecast_period]
        cache_quotes= cache_quotes.append(tmp)
    
    #根据未来振幅划分5个振动等级
    for i in cache_quotes.index:
        if cache_quotes.ix[i,['amplitude']].values>=0.01:
            cache_quotes.ix[i,['amplitude']]=1
        else:
            cache_quotes.ix[i,['amplitude']]=-1

    #获取股票多因子数据
    cache_fac= pd.DataFrame()
    for idate in trading_date_list:
        try:
            fac= DataAPI.MktStockFactorsOneDayGet(tradeDate=idate, secID=universe, field=["secID","tradeDate"]+used_factors, pandas="1").set_index('secID',drop=False)
            for ifac in used_factors:
                fac[ifac]=pd.Series(standardize(neutralize(winsorize(fac[ifac].to_dict(),n_draw=5),idate.strftime('%Y%m%d')))) 
            cache_fac= cache_fac.append(fac)
        except:
            print 'wrong in',idate
    cache= pd.merge(cache_quotes, cache_fac)    
    return cache
```

```python
# 随机抽取的数据集无任何金融、经济、计量、统计等方面的考虑。
universe= set_universe('ZZ500','2014-01-01')
l1 = ['EMA12','EMA26', 'VEMA10','VEMA26','PSY','RSTR12','WVAD','REC','Volatility','Skewness', 'PSY','ROA','PE', 'LCAP','BIAS10','ATR14']
tmp = SNW_Quotes_Amplitude_Classification(universe=universe, used_factors= l1,start_date='20140101', end_date='20141230', forecast_period=8)
tmp = tmp.dropna()
tmp.to_csv('tmp.csv')
```

## 集成算法框架
### 均值
- Bagging meta-estimator
- bagging通过对基于原始训练集的随机样本的黑盒学习器的结果进行整合，组成最后的预测结果。
- 通过使用在构建过程中的随机化来减少基学习器的方差，然后通过投票或平均做出最后的预测。
- bagging不用必须适配每一个学习器，它提供了减少过拟合的方法。

```python
from sklearn.linear_model import LogisticRegression #逻辑回归
from sklearn.naive_bayes import GaussianNB # 朴素贝叶斯 正态分布 这里分类划分不符合正态分布
from sklearn.ensemble import RandomForestClassifier #随机森林
from sklearn.svm import SVC #向量机
from sklearn.tree import DecisionTreeClassifier # 决策树
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import AdaBoostClassifier #迭代算法
from sklearn.ensemble import BaggingClassifier

LR = LogisticRegression(solver = 'newton-cg')
GNB = GaussianNB()
RFC = RandomForestClassifier(n_estimators = 200,max_depth=8,min_samples_leaf=10)
SM = SVC(C=10,cache_size=2000,gamma=1.0,decision_function_shape='ovo', probability=True)
DT = DecisionTreeClassifier(max_depth=10,min_samples_leaf=5)
GBC = GradientBoostingClassifier(n_estimators=200, learning_rate=1.0, max_depth=1, random_state=0)
AdaR_DT=AdaBoostClassifier(base_estimator=DecisionTreeClassifier(max_depth=10,min_samples_leaf=5)
                           ,learning_rate=1.0,n_estimators= 200,algorithm='SAMME.R')
```

```python
BR = BaggingClassifier(base_estimator=LR , n_estimators=100)
BR.fit(trainX,trainY)
print '拟合错误', 1.0- BR.score(trainX,trainY)
print '泛化', 1.0 -BR.score(testX,testY)
```

```python
BR = BaggingClassifier(base_estimator=RFC , n_estimators=100)
BR.fit(trainX,trainY)
print '拟合错误', 1.0- BR.score(trainX,trainY)
print '泛化', 1.0 -BR.score(testX,testY)
```

### 迭代 调权
- Adaboost是一种迭代算法，其核心思想是针对同一个训练集训练不同的分类器(弱分类器)，然后把这些弱分类器集合起来，构成一个更强的最终分类器(强分类器)。
- 其算法本身是通过改变数据分布来实现的，它根据每次训练集之中每个样本的分类是否正确，以及上次的总体分类的准确率，来确定每个样本的权值。
- 将修改过权值的新数据集送给下层分类器进行训练，最后将每次训练得到的分类器最后融合起来，作为最后的决策分类器

```python
n_estimators = 1000
AdaR_DT=AdaBoostClassifier(base_estimator=DT,learning_rate=1.0,n_estimators=n_estimators,algorithm='SAMME.R')
AdaR_DT.fit(trainX,trainY)
fig = plt.figure(figsize=(14,5))
ax = fig.add_subplot(111)
AdaR_DT_err = np.zeros((n_estimators,))
for i, predictY in enumerate(AdaR_DT.staged_predict(trainX)):
    AdaR_DT_err[i] = zero_one_loss(predictY, trainY)
PAdaR_DT_err = np.zeros((n_estimators,))
for i, ppredictY in enumerate(AdaR_DT.staged_predict(testX)):
    PAdaR_DT_err[i] = zero_one_loss(ppredictY, testY)
ax.plot(np.arange(n_estimators) + 1, AdaR_DT_err,label='Discrete AdaBoost Test Error',color='red')
ax.plot(np.arange(n_estimators) + 1, PAdaR_DT_err,label='Discrete AdaBoost Test Error',color='green')
print 'Adaboost 错误率',1-AdaR_DT.score(trainX,trainY)
print 'Adaboost 预测错误率',1-AdaR_DT.score(testX,testY)
```

```python
RFC = RandomForestClassifier(n_estimators = 100,max_depth=10,min_samples_leaf=2)
RFC.fit(trainX,trainY)
RFC_err=1.0 - RFC.score(trainX,trainY)
print '决策树错误率',RFC_err
print '决策树预测错误率', 1.0 - RFC.score(testX,testY)
#------------------------------------------------------------------------------------------------------------------------------------------------------
n_estimators = 50
AdaR_RFC=AdaBoostClassifier(base_estimator=RFC,learning_rate=1.0,n_estimators=n_estimators,algorithm='SAMME.R')
AdaR_RFC.fit(trainX,trainY)
fig = plt.figure(figsize=(14,5))
ax = fig.add_subplot(111)
AdaR_RFC_err = np.zeros((n_estimators,))
for i, predictY in enumerate(AdaR_RFC.staged_predict(trainX)):
    AdaR_RFC_err[i] = zero_one_loss(predictY, trainY)
PAdaR_RFC_err = np.zeros((n_estimators,))
for i, ppredictY in enumerate(AdaR_RFC.staged_predict(testX)):
    PAdaR_RFC_err[i] = zero_one_loss(ppredictY, testY)
ax.plot(np.arange(n_estimators) + 1, AdaR_RFC_err,label='Discrete AdaBoost Test Error',color='red')
ax.plot(np.arange(n_estimators) + 1, PAdaR_RFC_err,label='Discrete AdaBoost Test Error',color='green')
print 'Adaboost 错误率',1-AdaR_RFC.score(trainX,trainY)
print 'Adaboost 预测错误率',1-AdaR_RFC.score(testX,testY)
```

### 投票 VotingClassifier
- hard 少数服从多数，多数算法的民主专政！

```python
from sklearn.linear_model import LogisticRegression #逻辑回归
from sklearn.naive_bayes import GaussianNB # 朴素贝叶斯 正态分布 这里分类划分不符合正态分布
from sklearn.ensemble import RandomForestClassifier #随机森林
from sklearn.svm import SVC #向量机
from sklearn.tree import DecisionTreeClassifier # 决策树
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import AdaBoostClassifier #迭代算法
from sklearn.ensemble import VotingClassifier #投票算法
```

```python
LR = LogisticRegression(solver = 'newton-cg')
GNB = GaussianNB()
RFC = RandomForestClassifier(n_estimators = 200,max_depth=8,min_samples_leaf=10)
SM = SVC(C=10,cache_size=2000,gamma=1.0,decision_function_shape='ovo', probability=True)
DT = DecisionTreeClassifier(max_depth=10,min_samples_leaf=5)
GBC = GradientBoostingClassifier(n_estimators=200, learning_rate=1.0, max_depth=1, random_state=0)
AdaR_DT=AdaBoostClassifier(base_estimator=DecisionTreeClassifier(max_depth=10,min_samples_leaf=5)
                           ,learning_rate=1.0,n_estimators= 200,algorithm='SAMME.R')
```

```python
VC = VotingClassifier(estimators=[('LR',LR),('GNB',GNB),('RFC',RFC),('SVC',SM),('DT',DT),('GBC',GBC),('Ada',AdaR_DT)],voting='hard') #少数服从多数
VC.fit(trainX,trainY)
print '拟合错误', 1.0- VC.score(trainX,trainY)
print '泛化', 1.0 -VC.score(testX,testY)
```

### Bagging
- 1.给定一个弱学习算法,和一个训练集;
- 2.单个弱学习算法准确率不高;
- 3.将该学习算法使用多次,得出预测函数序列,进行投票;
- 4.最后结果准确率将得到提高.

```python
from sklearn.linear_model import LogisticRegression #逻辑回归
from sklearn.naive_bayes import GaussianNB # 朴素贝叶斯 正态分布 这里分类划分不符合正态分布
from sklearn.ensemble import RandomForestClassifier #随机森林
from sklearn.svm import SVC #向量机
from sklearn.tree import DecisionTreeClassifier # 决策树
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import AdaBoostClassifier #迭代算法
from sklearn.ensemble import VotingClassifier #投票算法

LR = LogisticRegression(solver = 'newton-cg')
GNB = GaussianNB()
RFC = RandomForestClassifier(n_estimators = 200,max_depth=8,min_samples_leaf=10)
SM = SVC(C=10,cache_size=2000,gamma=1.0,decision_function_shape='ovo', probability=True)
DT = DecisionTreeClassifier(max_depth=10,min_samples_leaf=5)
GBC = GradientBoostingClassifier(n_estimators=200, learning_rate=1.0, max_depth=1, random_state=0)
AdaR_DT=AdaBoostClassifier(base_estimator=DecisionTreeClassifier(max_depth=10,min_samples_leaf=5)
                           ,learning_rate=1.0,n_estimators= 200,algorithm='SAMME.R')
VC = VotingClassifier(estimators=[('LR',LR),('GNB',GNB),('RFC',RFC),('SVC',SM),('DT',DT),('GBC',GBC),('Ada',AdaR_DT)],voting='hard') #少数服从多数
```

```python
from sklearn.ensemble import BaggingClassifier
BC = BaggingClassifier(base_estimator=VC) #使用默认10次
BC.fit(trainX,trainY)
print '拟合错误', 1.0- BC.score(trainX,trainY)
print '泛化', 1.0 -BC.score(testX,testY)
```
