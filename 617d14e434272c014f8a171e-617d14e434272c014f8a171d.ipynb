# 行业成交量择时-申万一级行业

# 原理
    单个信号（均为大于触发）
    ｜ 成交量60百分位
    ｜ 成交额60百分位
    ｜ 月度成交量乖离率60百分位
    ｜ 月度成交额乖离率60百分位
    复合因子
    ｜ 所有4个因子中触发N个以上（不含N）信号作为卖出信号，否则买入或持有。N=2最好

```python
from lib.functions import *
start_date='20100101'
end_date='20211001'
```

# Step1:获取所有数据

```python
# 获取申万二级行业
indexSymbol=list(DataAPI.IndustryGet(industryVersionCD=u"010303",industryLevel=u"1",
                                        isNew=1,field=u"indexSymbol",pandas="1")['indexSymbol'])
# 获取所有数据
data=DataAPI.MktIdxmGet(ticker=indexSymbol,beginDate=start_date,endDate=end_date)
# 获取期间月末交易日列表
cal=list(data.endDate.drop_duplicates().sort_values(ascending=True))
#基准收益（所有行业等权配置）
base=data[['endDate','chgPct']].dropna().groupby('endDate').mean()
base.columns=['benchmark']
```

# Step2：成交量指标计算

## 计算成交量数据

```python
#成交量数据
turnover=data[['ticker','endDate','turnoverVol','turnoverValue']]#取得成交量、成交金额
turnover['to_bias']=turnover['turnoverVol']/turnover['turnoverVol'].rolling(12).mean()#计算成交量乖离率
turnover['ta_bias']=turnover['turnoverValue']/turnover['turnoverValue'].rolling(12).mean()#计算成交金额乖离率
```

## 汇总所有指数每月成交量数据便于下一步统计

```python
turnover=turnover.groupby(['ticker','endDate']).mean().unstack().T
#用字典方便循环
turnovers={}
keys=['turnoverVol','turnoverValue','to_bias','ta_bias']
for k in keys:
    turnovers[k]=turnover.loc[k]
```

## 计算当月指标（是否大于12个月的60百分位）

```python
#计算每月指标
crowd={}
for k in keys:
    q=turnovers[k].rolling(12).quantile(0.9)#12个月滚动60百分位
    crowd[k]=turnovers[k]>q
```

# setp3：单个行业择时结果和行业指数对比函数

```python
import numpy as np
def compare(indexSymbol):
    # 获取行业指数
    industry_index=DataAPI.MktIdxmGet(ticker=indexSymbol,beginDate=start_date,endDate=end_date,field='endDate,closePrice')
    industry_index.dropna(inplace=True)
    industry_index.set_index('endDate',inplace=True)
    industry_index.sort_index(ascending=True,inplace=True)
    # 起点指数设为1
    industry_index=industry_index/industry_index.iloc[0]
    result=industry_index[:]#记录对比结果
    tmp_rtn={}#记录每月择时后净值
    for i in range(0,len(cal)-1):
        # 获取成交量记录
        turnoverValue=crowd['turnoverValue'][indexSymbol].loc[cal[i]]#成交额
        turnoverVol=crowd['turnoverVol'][indexSymbol].loc[cal[i]]#成交量
        to_bias=crowd['to_bias'][indexSymbol].loc[cal[i]]#成交量乖离率
        ta_bias=crowd['ta_bias'][indexSymbol].loc[cal[i]]#成交额乖离率
        signal=np.sum([turnoverValue,turnoverVol,to_bias,ta_bias])
        if signal>2 or (cal[i] not in industry_index.index or cal[i+1] not in industry_index.index):
            tmp_rtn[cal[i+1]]=0
        else:
            tmp_rtn[cal[i+1]]=(industry_index.loc[cal[i+1]]/industry_index.loc[cal[i]]-1).values[0]
    result['return']=(pd.Series(tmp_rtn)+1).cumprod()
    result.dropna(inplace=True)
    return result
```

# step4：测试所有申万二级行业效果

```python
#所有二级行业测试效果
summary={}
industry=DataAPI.IndustryGet(industryVersionCD=u"010303",industryLevel=u"1",
                                        isNew=1,field=u"indexSymbol,industryName",pandas="1")
for ind in list(industry.indexSymbol):
    rtn=compare(ind)
    ind_name=industry[industry.indexSymbol==ind]['industryName'].values[0]
    ann_rtn=pow(rtn.iloc[-1],12.0/len(rtn))-1#计算年化收益率
    max_down=(1-rtn/rtn.cummax()).max()#计算最大回撤
    summary[ind_name]={}
    summary[ind_name]['benchmark_return']=ann_rtn.iloc[0]#基准年化收益率
    summary[ind_name]['selection_return']=ann_rtn.iloc[1]#择时年化收益率
    summary[ind_name]['benchmark_maxdown']=max_down.iloc[0]#基准最大回撤
    summary[ind_name]['selection_maxdown']=max_down.iloc[1]#择时最大回撤
```
