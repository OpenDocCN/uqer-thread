# 优矿杜邦分析 点：分类求和平均；分位数与计数

```python
##杜邦分析
df=DataAPI.FdmtIndiDupontGet(ticker=u"000425",secID="",endDate="",beginDate="",beginYear=u"2011",endYear=u"2013",reportType=u"A",field=u"endDate,ROE,npMargin,taTurnover,equMultiplierDupont,finanExp,incomeTax",pandas="1")
bs=DataAPI.FdmtBSGet(ticker=u"000001",secID=u"",reportType=u"A",endDate=u"",beginDate=u"",publishDateEnd=u"",publishDateBegin=u"",endDateRep="",beginDateRep="",beginYear="2011",endYear="2013",fiscalPeriod="",field=u"endDate,TAssets,TShEquity",pandas="1")
income=DataAPI.FdmtISGet(ticker=u"000002",secID=u"",reportType=u"A",endDate=u"",beginDate=u"",publishDateEnd=u"",publishDateBegin=u"",endDateRep="",beginDateRep="",beginYear="2011",endYear="2013",fiscalPeriod="",field=u"endDate,NIncome,TProfit,incomeTax,tRevenue",pandas="1")
ebit=DataAPI.FdmtDerGet(secID="",ticker=u"000425",beginYear=u"2011",endYear=u"2013",reportType=u"A",beginDate="",endDate="",field=u"endDate,EBIT",pandas="1")
df=df.merge(bs,left_on=['endDate'],right_on=['endDate'],how='inner')
df=df.merge(income,left_on=['endDate'],right_on=['endDate'],how='inner')
df=df.merge(ebit,left_on=['endDate'],right_on=['endDate'],how='inner').drop_duplicates(subset=['endDate'], keep='first')
df['ssfd']=df['NIncome']/df['TProfit']
df['lxfd']=df['TProfit']/df['EBIT']
df['grossmargin']=df['EBIT']/df['tRevenue']
df['turnover']=df['tRevenue']/df['TAssets']
df['ggl']=df['TAssets']/df['TShEquity']
print df[['endDate','ROE','ssfd','lxfd','grossmargin','turnover','ggl']]
print df[['endDate','ROE','npMargin','taTurnover','equMultiplierDupont']]
```
