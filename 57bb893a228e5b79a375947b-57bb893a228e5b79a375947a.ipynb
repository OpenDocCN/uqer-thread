```python
import pandas as pd
from pandas import DataFrame, Series
universe=set_universe('HS300')
startdate = '20160101'
enddate = '20161231'
rtns = DataFrame()
for i in range(len(universe)):
    cp=DataAPI.MktEqudAdjGet(secID=str(universe[i]),isOpen="1",beginDate=startdate,endDate=enddate,field=u"tradeDate,ticker,closePrice",pandas="1")
    cp.sort(columns = 'tradeDate', inplace = True)
    cp.columns = ['ticker','tradeDate','return']
    cp['return'][1:] = 100.0 * (cp['return'][1:].values / cp['return'][:1].values - 1)
    cp['return'][:1] = 0
    rtns = pd.concat([rtns,cp],axis = 0)  #  dataframe连接操作
rtn_table = pd.crosstab(rtns['tradeDate'],rtns['ticker'], values = rtns['return'], aggfunc = sum)  #  一维表变为二维表
rtn_table.fillna(0, inplace = True)  #  将NaN置换为0
rtn_table.to_csv('tmp.csv')
rtn_table=rtn_table.T
rtn_table.plot(figsize=(30, 15),legend=False)
```
