```python
def maxdd(df,col_val,col_date):
    """
    maxdrawdown of a security
    :param df: DataFrame of the historical stock data
    :param col_val: column name of df,价格历史数据
    :param col_date: column name of df,日期历史数据
    :return: 
    """
    if not df.empty:
        res = df[[col_date,col_val]]
        res['rel_max'] = pd.expanding_max(res[col_val])
        res['drawdown'] = 1 - res[col_val] / res['rel_max']
        maxdd = res['drawdown'].max()

        idx_bottom = res['drawdown'].argmax()
        date_bottom = res.ix[idx_bottom][col_date]  # 最大回撤的低点

        idx_top = res.loc[:idx_bottom, col_val].argmax()
        date_top = res.ix[idx_top][col_date]  # 最大回撤的高点
        return date_top,date_bottom,maxdd
    else:
        return 0,0,0
```

```python
# 计算最大回撤
historical_data = DataAPI.FundNavGet(dataDate=u"",secID=fundID,ticker=u"",beginDate=u"",endDate=u"",field=u"",pandas="1")
historical_data['endDate'] = pd.to_datetime(historical_data['endDate'])
begin,end,maxdrawdown = maxdd(historical_data,'ADJUST_NAV','endDate')
print(begin,end,maxdrawdown)
```

```python
# 最大回撤可视化
ax= historical_data.plot('endDate','ADJUST_NAV',figsize=(18,9),legend=False)
historical_data[(historical_data['endDate'] > begin) & (historical_data['endDate']<end)].plot('endDate','ADJUST_NAV',ax=ax,color='red',legend=False,grid=True)
plt.show()
```

```python
data = historical_data[['endDate','ADJUST_NAV']]
monthly_return = data.set_index('endDate').resample('MS').first().pct_change()
volatility = data.set_index('endDate').resample('MS').std()
```

```python
monthly_return = monthly_return.rename(columns={'ADJUST_NAV':'ret'})
volatility = volatility.rename(columns={'ADJUST_NAV':'std'})
df = monthly_return.join(volatility)
df[df.index > '2015'].plot(kind='bar',figsize=(18,12))
plt.show()
```
