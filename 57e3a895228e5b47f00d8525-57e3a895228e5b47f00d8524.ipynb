让我们来看以上的代码。
根据[常用API文档](https://uqer.datayes.com/help/faqApi/)，```initialize(account):```和```handle_data(account):```在回测期的每个交易日会被调用一次。因此为了验证这一点，在上段代码的```handle_data(account):```中加入了```print '---the day begins---'```和```print '---end of this day---'```。选择2013-01-01到2013-01-07，按照上交所的日历，其中包含两个工作日，因此从```print```函数的输出情况来看，```handle_data(account):```被调用了两次。

Python内置的datetime模块中的时间对象与CAL中Date对象的区别，通过相应的```print```函数的输出结果，可略知一二。由于周末、节假日以及两个交易所不调休的原因，对于日期涉及到工作日的判断，这也就是原作为何引入了CAL中的相关模块。

quartz框架能够接收的日期数据结构为```'20160923'```或```'2016-09-23'```，所以调用了CAL中Date对象的```strftime()```方法将Date对象转化为quartz能够接收的日期数据结构。

[常用API文档](https://uqer.datayes.com/help/faqApi/)告诉我们```set_universe()```函数支持返回成分股代码列表，从```print```函数的输出情况来看，也的确输出了“HS300”沪深300的成分股代码。


为了方便阅读，以上代码中删去了已经重复部分的```print```代码，在本作的最后，会将所有作为“断点”的```print```语句加入代码作为完整版进行呈现。

通过```DataAPI.MktStockFactorsOneDayGet()```函数返回了证券池中股票的相关信息，并定义了一个叫```getData_current_date```的变量指向这些返回的信息。```getData_current_date```的数据类型为pandas中的DataFrame，可以简单地理解为一个二维dict。不过美中不足的是，```getData_current_date```的行标（index）并不是证券池中股票的交易代码，而是打印出来的```[0, 1, 2, 3...]```，但是我们希望```getData_current_date```的行标（index）是```['000883.XSHE', '600895.XSHG'...]```这样的证券交易代码。因此，下一行代码调用了DataFrame对象的```set_index()```方法来修改行标（index）。在up主[上一篇](https://uqer.datayes.com/community/share/57e27d04228e5b049cfb9692)帖子中，对于```set_index()```有所讲解。

```getData_last_date```的作用与```DataAPI.MktStockFactorsOneDayGet()```相同，唯一的区别在于返回上一个交易日的证券池中股票的相关信息。


```getData_current_date=getData_current_date[getData_current_date.NetProfitGrowRate>=1.0].dropna()```作用是将证券池中```NetProfitGrowRate>=1.0```的股票筛选出来。

```for stock in list(getData_current_date.index):```遍历```getData_current_date.index```中的元素。假设```getData_current_date.index```中有10只股票，那么这个循环体将会循环10次，从循环体中的```print```语句的输出情况来看，的确是这样的。

```(getData_current_date['MA5'][stock]>getData_current_date['MA10'][stock])&(getData_last_date['MA5'][stock]<=getData_last_date['MA10'][stock])```表示“5日均线上传10日均线”的代码，当证券池中的某只股票符合这种条件之后，会被加入```buylist```中，作为提示，在触发```buylist.append(stock)```之前，会首先触发```print 'this is buylist'```作为提示。涉及```selist```的语句有着相似的原理，所以在第一个交易日的循环中，```buylist```没有被触发，而```selist```被触发属于正常的情况，因为在后面卖出的时候，判断依据并不是证券要先在```buylist```而是要在```account.valid_secpos```当中，这一点后面会有所展开。

胜利在望！

[常用API文档](https://uqer.datayes.com/help/faqApi/)告诉我们```account.valid_secpos```表示上一个K线结束后有效持仓量，只有目前持仓的股票才可以卖出，因此卖出指令```order_to(stock, 0)```的判断条件有两个：1）5日线下穿10日线，也就是循环体的判断条件```stock in selist```；2）是目前正在持仓的股票，也就是```if```判断条件```stock in account.valid_secpos```。

```order```和```order_to```在执行之后会当天买入吗？答案是否定的，由于原作采用的是日间策略```freq='d'```且```refresh_rate=1```所以在当天触发买入和卖出指令之后，并不会买进和卖出，而是将买入和卖出的命令传入```account.blotter```当中。这段回测时间内的第三个交易日，从打印结果可以看到，产生了第一个买入指令：买入“002008.XSHE”12500股，但是在```order(stock, account.cash/account.referencePrice[stock]/len(buylist)```之后的```print accout.valid_secpos```的输出结果却是```{}```，同时触发```order()```之前，打印结果显示，```account.valid_secpos```也是```{}```，说明没有持仓，也就是说并没有买入，而是将买入指令存到了```account.blotter```中。
