# 协整性

前面看了高人的《基于时间序列的协整关系的配对交易》受益匪浅，但是仔细研究了一下协整性，发现有些地方错误，或者说是我的理解不对，所以重新写一个，抛砖引玉。

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import seaborn as sns
import matplotlib.pyplot as plt

def find_cointegrated_pairs(dataframe):
    ii=0
    n = dataframe.shape[1]
    pvalue_matrix = np.ones((n, n))
    keys = dataframe.keys()
    pairs = []
    for i in range(n):
        for j in range(i+1, n):
            stock1 = dataframe[keys[i]]
            stock2 = dataframe[keys[j]]
            ii+=1
            result = sm.tsa.stattools.coint(stock1, stock2)
            pvalue = result[1]
            pvalue_matrix[i, j] = pvalue
            if pvalue < 0.05:
                pairs.append((keys[i], keys[j], pvalue))
    return pvalue_matrix, pairs

data=DataAPI.SecTypeRelGet(typeID=u"101001002018001001",secID=u"",ticker=u"",field=u"",pandas="1")
stock_list = np.array(data.secID)
#stock_list = list(set(stock_list))

df1=pd.DataFrame()
prices_df=pd.DataFrame()
for stock in stock_list:
    data=DataAPI.MktEqudAdjGet(secID=stock,ticker=u"",isOpen="",beginDate=u"20150101",endDate=u"20160101",pandas="1")
    df1=pd.DataFrame(data.closePrice.values, index=[data.tradeDate], columns=[stock])
    prices_df=pd.concat([prices_df,df1],axis=1)

pvalues, pairs = find_cointegrated_pairs(prices_df)
sns.heatmap(1-pvalues, xticklabels=stock_list, yticklabels=stock_list, cmap='RdYlGn_r', mask = (pvalues == 1))
print pairs
```

```python
y=1.45*stock1-stock2
y.plot();
plt.axhline((y).mean(), color="red", linestyle="--")
plt.xlabel("Time"); plt.ylabel("Stationary Series")
plt.legend(["price spreed", "Mean"])
```

怎么衡量价差超过多少我们就进行配对交易呢，引入一个z score的概念。Z值的量代表着原始分数和母体平均值之间的距离，是以标准差为单位计算。
但是实盘有待验证的问题：这里的均值和标准差是用2016年1月1日到2017年2月1日的收盘价算出来的，z score的数值也是用这部分时间的数据来算的。是未来函数。

```python

def zscore(series):
    return (series - series.mean()) / np.std(series)

zs=zscore(1.45*stock1-stock2)
zs.plot()

plt.axhline(zs.mean(), color="black")
plt.axhline(1.0, color="magenta", linestyle="--")
plt.axhline(-1.0, color="cyan", linestyle="--")
plt.legend(["z-score", "mean", "+1", "-1"])
```

由上图可以看出，总体当z score超过1和-1的时候，还是会往均值走的。

那么我们现在用2016年到2017年的数据验证一下策略的操作。

```python
import matplotlib.pylab as plt

start = '2016-01-01'                       # 回测起始时间
end = '2017-02-01'   
stock1 = DataAPI.MktEqudGet(secID='601398.XSHG',ticker=u"",beginDate=start,endDate=end,isOpen="",field=u"",pandas="1")
stock2 = DataAPI.MktEqudGet(secID='601939.XSHG',ticker=u"",beginDate=start,endDate=end,isOpen="",field=u"",pandas="1")
stock1=stock1[['tradeDate','closePrice']]
stock1.set_index('tradeDate',inplace=True)
stock2=stock2[['tradeDate','closePrice']]
stock2.set_index('tradeDate',inplace=True)

stock1.closePrice.plot()
stock2.closePrice.plot()
plt.xlabel("Time"); plt.ylabel("Price")
plt.legend(["601398", "601939"],loc='best')
```

```python
x = stock1.closePrice
x.name='601398.XSHG'
y = stock2.closePrice
X = sm.add_constant(x)
result = (sm.OLS(y,X)).fit()
print(result.summary())
# formula = 1.75*'601398.XSHG' - '601939.XSHG'
```

```python
def zscore(series):
    return (series - series.mean()) / np.std(series)

zs=zscore(1.45*stock1-stock2).closePrice
zs.plot()

plt.axhline(zs.mean(), color="black")
plt.axhline(1.0, color="magenta", linestyle="--")
plt.axhline(-1.0, color="cyan", linestyle="--")
plt.legend(["z-score", "mean", "+1", "-1"])
```

用2016年到2017年内的数据看了一下z score，然后去对比了一下持仓，比如在2016年7月份，z score大于1，持仓应该换到建设银行。策略的执行没有问题。那么策略还是亏损的。我对比的benchmark是工商银行。
