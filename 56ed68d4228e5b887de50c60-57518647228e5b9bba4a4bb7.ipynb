```python
import pandas as pd
from pandas import DataFrame
import time
import re
data=pd.read_csv('top.csv')
data.head()
data=data.iloc[:,:-2]

id=list(data['id'])

dates=DataAPI.TradeCalGet(exchangeCD=u"XSHG",beginDate=u"20160104",endDate=u"20160311",field=u"",pandas="1")
dates=dates[dates.isOpen==1]
opendays=list(dates['calendarDate'])

def consumer(date,date2):
    #修改查询日期格式
    #date='2016-03-03'
    timestamp=time.strptime(date,'%Y-%m-%d')
    b=time.strptime(date2,'%Y-%m-%d')
    a=int(time.mktime(timestamp))-24*3600
    a=time.localtime(a)
    daytime1=time.strftime("%b%d", timestamp)
    daytime2=time.strftime("%b%d", b)
    change=data[daytime1]-data[daytime2]
    change=DataFrame(list(change),index=id)
    change.columns=['followers']
    top=change.sort(columns='followers',ascending=False)[:20]
    return top
collection=[]
for i in range(len(opendays)):
    if i==0:
        print ('开始计算')
    else:
        top=consumer(opendays[i],opendays[i-1])
        collection.append(top)

def gettop(i):
    codes=[]
    for code in list(collection[i].index):        
        a=re.search('\d{6}',code).group(0)
        if int(a)>400000:
            a=str(a)+'.XSHG'
        else:
            a=str(a)+'.XSHE'
        codes.append(a)
    return codes
gettop(0)
```
