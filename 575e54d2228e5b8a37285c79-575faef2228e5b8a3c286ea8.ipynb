```python
import talib as ta 
from CAL.PyCAL import *
import numpy as np
import pandas as pd


def shizhizuixiao(account):  #判断条件2
    shizhi=[]
    shizhidt=DataAPI.MktStockFactorsOneDayGet(tradeDate=account.previous_date,secID=account.universe,field=u"secID,LCAP",pandas='1')
    shizhidt=shizhidt.dropna()
    for stock,LCAP in zip(shizhidt['secID'].values,shizhidt['LCAP'].values):
        shizhi.append((stock,LCAP))
    shizhi=sorted(shizhi,key=lambda k:k[-1],reverse=False)        #按市值排序
    shizhi=shizhi[:100]                                         #选出最前面的100只
    shizhidict=dict(shizhi)
    a=shizhidict.keys()                                        #保留股票代码
    return a
def zhangdieting(stocklist,date):  #判断条件3
    zdt=[]
    if stocklist is None or len(stocklist)==0:
        return zdt
    df_price = DataAPI.MktEqudGet(tradeDate=date,secID=stocklist,field=u"secID,closePrice",pandas="1")
    df_upLimit = DataAPI.MktLimitGet(secID=stocklist,tradeDate=date,field=u"secID,limitUpPrice,limitDownPrice",pandas="1")
    df_price=df_price.dropna()
    df_upLimit=df_upLimit.dropna()
    df=df_price.merge(df_upLimit)
    for stock,closeprice,up,down in zip(df['secID'].values,df['closePrice'].values,df['limitUpPrice'].values,df['limitDownPrice'].values):
        if closeprice==up or closeprice==down:   
            pass
        else:
            zdt.append(stock)   #如果昨天没涨跌停，记录下来
    return zdt

def ma5ma10(stocklist,date):  #判断条件4
    duotou=[]
    if stocklist is None or len(stocklist)==0:
        return duotou

    junxian = DataAPI.MktStockFactorsOneDayGet(tradeDate=date,secID=stocklist,field=['secID','MA5','MA10'],pandas="1")
    junxian=junxian.dropna()
    for stock, ma5, ma10 in zip(junxian['secID'].values, junxian['MA5'].values, junxian['MA10'].values):
        if ma5>ma10:
            duotou.append(stock)
    return duotou

def rsipd(stocklist,date):
    rsipd=[]
    ####get the daily data
    cal = Calendar('China.SSE')
    last40day = cal.advanceDate(date,'-40B',BizDayConvention.Preceding).strftime("%Y%m%d") 
    for stock in stocklist:
        spdb = DataAPI.MktEqudAdjGet(secID=stock,beginDate=last40day,endDate=date,field='tradeDate,secID,closePrice')
    ####change the index of the data
        spdb = spdb.set_index('tradeDate').sort_index()
        spdb.index = pd.to_datetime(spdb.index)
    ####timeperiod = 14
        rsik = ta.RSI(spdb.closePrice.values,12)
        rsim = ta.RSI(spdb.closePrice.values,36)
        if rsik[-1]> rsim[-1] and rsik[-1]>50:
            rsipd.append(stock)

    return rsipd

def dapan(account):   #判断大盘条件5，不满足返回True，满足条件5返回False
    date=account.previous_date
    cal = Calendar('China.SSE')
    last22day = cal.advanceDate(date,'-22B',BizDayConvention.Preceding).strftime("%Y%m%d") # 获取回测当日的前21天日期,
    info = DataAPI.MktIdxdGet(ticker='000905',beginDate=last22day,endDate=date,field='closeIndex')      #修改成了zz500的指数
    info.dropna()
    info=info['closeIndex'].values
    ma20_pre_pre=np.mean(info[-21:-2])
    ma20_pre=np.mean(info[-20:-1])
    ma5_pre=np.mean(info[-5:])
    closeindex=info[-1]
    if ma20_pre_pre>ma20_pre or closeindex <ma5_pre:
        dapan=False
    else:
        dapan=True
    return dapan
```
