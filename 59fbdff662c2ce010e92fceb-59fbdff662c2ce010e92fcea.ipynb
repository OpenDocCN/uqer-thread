# 计算年份数据

```python
import numpy as np
import pandas as pd
#查找年份的数据
year = ['2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017']
date = ['2008-12-31', '2009-12-31', '2010-12-31', '2011-12-30', '2012-12-31', '2013-12-31', '2014-12-31', '2015-12-31', '2016-12-30', '2017-11-02']
field = ['ticker', 'secShortName', 'tradeDate', 'openIndex', 'lowestIndex', 'highestIndex', 'closeIndex']

def returnTable(secID):
    #输入证券代码，返回年份数据    
    table = pd.DataFrame()
    table['ticker'] = None
    table['name'] = None
    table['year'] = year
    table['date'] = date
    table['close'] = None
    table['mean'] =None
    table['chg'] = None
    table['chgpct'] = None

    data = DataAPI.MktIdxdGet(tradeDate=u"",indexID=u"",ticker=secID,beginDate=u"2008-01-01",endDate=u"2017-12-31",exchangeCD=u"XSHE,XSHG",field=field,pandas="1")
    data['year'] = data['tradeDate'].astype(str).str[0:4]

    for idx in table.index:
        table.ix[idx, 'ticker'] = data.ix[0, 'ticker']
        table.ix[idx, 'name'] = data.ix[0, 'secShortName']
        table.ix[idx, 'mean'] = np.mean(data.ix[data['year'] == year[idx]]['closeIndex'])
        table.ix[idx, 'close'] = data.ix[data['tradeDate'] == table.ix[idx, 'date']]['closeIndex'].values[0]
    for idx in table.index[:-1]: 
        table.ix[idx+1, 'chg'] = table.ix[idx+1, 'close'] - table.ix[idx, 'close']
        table.ix[idx+1, 'chgpct'] = (table.ix[idx+1, 'close'] / table.ix[idx, 'close'] - 1) * 100
    table.columns = ['代码', '名称', '年份', '日期', '收盘价', '年均价', '年涨跌', '年涨跌幅（%）']
    return table
```
