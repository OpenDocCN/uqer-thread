# 满仓踏空

###满仓踏空：历史总是惊人地相似
http://3g.k.sohu.com/t/n27403847?from=singlemessage&isappinstalled=0

2014年10月28日到12月5日，中国股市迎来一波强劲的上涨，上证综指和深证成指的涨幅均超过了25%。然而很多股民却发现自己“满仓踏空”，十分令人困惑。

```python
from CAL.PyCAL import *

df_SZZZ_now = DataAPI.MktIdxdGet(indexID = "000001.ZICN", field = ["closeIndex"], beginDate = "20141028", endDate = "20141206")
df_SZCZ_now = DataAPI.MktIdxdGet(indexID = "399001.ZICN", field = ["closeIndex"], beginDate = "20141028", endDate = "20141206")

df_SZZZ_now['pct'] = df_SZZZ_now['closeIndex']/df_SZZZ_now['closeIndex'].iloc[0] - 1
df_SZCZ_now['pct'] = df_SZCZ_now['closeIndex']/df_SZCZ_now['closeIndex'].iloc[0] - 1
fig = pylab.figure(figsize = [10,5])
df_SZZZ_now['pct'].plot()
df_SZCZ_now['pct'].plot()
pylab.legend(["ShangZhengZongZhi", "ShenZhengChengZhi"], loc=2)
```

```python
"ShangZhengZongZhi return: %f" % (df_SZZZ_now['closeIndex'].iloc[-1]/df_SZZZ_now['closeIndex'].iloc[0] - 1)
```

```python
"ShenZhengChengZhi return: %f" % (df_SZCZ_now['closeIndex'].iloc[-1]/df_SZCZ_now['closeIndex'].iloc[0] - 1)
```

在这一轮上涨中，最抢眼的是银行股、券商股等金融股。金融股领跑大盘，从涨幅前20的股票来看，券商股占据了相当多的位置。

```python
a_stocklist_now = [r.strip() for r in file("A.csv").readlines()]
target_len_now = len(DataAPI.MktEqudGet(secID = "601398.XSHG", field = ["closePrice"], beginDate = "20141028", endDate = "20141206"))
target_ret_now = df_SZZZ_now['closeIndex'].iloc[-1]/df_SZZZ_now['closeIndex'].iloc[0] - 1

a_validlist_now, a_financial_now = {}, {}
for stk in a_stocklist_now:
    df_now = DataAPI.MktEqudGet(secID = stk, field = ["closePrice"], beginDate = "20141028", endDate = "20141206")
    ind_now = DataAPI.EquIndustryGet(secID = stk)
    if len(ind_now) != 1: continue
    
    ind_now = ind_now.at[0, 'industryName1']
    if len(df_now) == target_len_now and df_now['closePrice'].iloc[-1] != df_now['closePrice'].iloc[0]:
        a_validlist_now[stk] = df_now['closePrice'].iloc[-1]/df_now['closePrice'].iloc[0] - 1
        if ind_now in [u'银行', u'非银金融']:
            a_financial_now[stk] = df_now['closePrice'].iloc[-1]/df_now['closePrice'].iloc[0] - 1
```

```python
slist_now = sorted(a_validlist_now.keys(), key=lambda x: a_validlist_now[x], reverse=True)[:20]
df_top20_now = DataAPI.SecIDGet(ticker=",".join(r[:6] for r in slist_now), assetClass="E", field=["secShortName"])
df_top20_now["return"] = [a_validlist_now[x] for x in df_top20_now["secID"]]
df_top20_now = df_top20_now.loc[:,["secID", "secShortName", "return"]].sort(column = "return", ascending = False)
df_top20_now
```

50只金融股平均涨幅达到了惊人的65.8%！然而除去金融股之外，剩下的所有股票平均涨幅仅为7.5%，和大盘指数的25%+相距甚远。

```python
"Amount of financial stocks: %d" % len(a_financial_now)
```

```python
"Average return of financial stocks: %f" % (sum(a_financial_now[k] for k in a_financial_now)/len(a_financial_now))
```

```python
"Average return of stocks except financial sector: %f" % (sum(a_validlist_now[k] for k in a_validlist_now if k not in a_financial_now)/(len(a_validlist_now)-len(a_financial_now)))
```

纵观全局，也仅有287只股票跑赢大盘的25%收益率，占全部股票的12.1%。因此，“满仓踏空”不会只是个别现象。

```python
i = 0.
for x in a_validlist_now:
    if a_validlist_now[x] >= target_ret_now:
        i += 1
```

```python
"Amount of stocks whose return higher than ShangZhengZongZhi: %d" % i
```

```python
"Amount of stocks whose return higher than ShangZhengZongZhi: %f" % (i/len(a_validlist_now))
```

那么广大股民遇到现在的“满仓踏空”应该怎么办呢？我们不妨先回顾一下8年前的冬天，通过比较我们会发现，历史总是惊人地相似。

从2006年12月11日到2007年1月4日，在2007年上半年那一波前所未有的上涨来临之前，中国同样经历了大盘拉升，但股民却“满仓踏空”的过程。在那段时间里，上证综指涨幅达24.5%，而深证成指涨幅也达到了12.6%。

```python
df_SZZZ = DataAPI.MktIdxdGet(indexID = "000001.ZICN", field = ["closeIndex"], beginDate = "20061211", endDate = "20070104")
df_SZCZ = DataAPI.MktIdxdGet(indexID = "399001.ZICN", field = ["closeIndex"], beginDate = "20061211", endDate = "20070104")

df_SZZZ['pct'] = df_SZZZ['closeIndex']/df_SZZZ['closeIndex'].iloc[0] - 1
df_SZCZ['pct'] = df_SZCZ['closeIndex']/df_SZCZ['closeIndex'].iloc[0] - 1

fig = pylab.figure(figsize = [10,5])
df_SZZZ['pct'].plot()
df_SZCZ['pct'].plot()
pylab.legend(["ShangZhengZongZhi", "ShenZhengChengZhi"], loc=2)
```

沪市涨幅前5的股票中，中国银行和工商银行分列第二和第四。

```python
a_stocklist = [r.strip() for r in file("A.csv").readlines()]
target_len = len(DataAPI.MktEqudGet(secID = "601398.XSHG", field = ["closePrice"], beginDate = "20061210", endDate = "20070104"))
df_SZZZ = DataAPI.MktIdxdGet(indexID = "000001.ZICN", field = ["closeIndex"], beginDate = "20061210", endDate = "20070104")
target_ret = df_SZZZ['closeIndex'].iloc[-1]/df_SZZZ['closeIndex'].iloc[0] - 1

a_validlist, a_financial = {}, {}
for stk in a_stocklist:
    df = DataAPI.MktEqudGet(secID = stk, field = ["closePrice"], beginDate = "20061210", endDate = "20070104")
    ind = DataAPI.EquIndustryGet(secID = stk)
    if len(ind) != 1: continue
    
    ind = ind.at[0, 'industryName1']
    if len(df) == target_len and df['closePrice'].iloc[-1] != df['closePrice'].iloc[0]:
        a_validlist[stk] = df['closePrice'].iloc[-1]/df['closePrice'].iloc[0] - 1
        if ind in [u'银行', u'非银金融']:
            a_financial[stk] = df['closePrice'].iloc[-1]/df['closePrice'].iloc[0] - 1
```

```python
slist = sorted([r for r in a_validlist.keys() if r[0]=='6'], key=lambda x: a_validlist[x], reverse=True)[:5]
df_top5 = DataAPI.SecIDGet(ticker=",".join(r[:6] for r in slist), assetClass="E", field=["secShortName"])
df_top5["return"] = [a_validlist[x] for x in df_top5["secID"]]
df_top5 = df_top5.loc[:,["secID", "secShortName", "return"]].sort(column = "return", ascending = False)
df_top5
```
