结合 FAMA在2015年写的论文，我们借助其五因子模型对A股市场进行实证检验，其中在具体存在争议的方法和定义上统一做以下规定：

    1. RMW:营业收益率因子，即为ROE（净资产收益率）分组差值
    2.CMA：投资水平，（△资产/（去年资产））分组差值
    3.HML：1/PB分组差值
    4.SMB:SIZE分组差值
    5.无风险利率选择为shibor隔夜利率
    6.因子分组数量和方法，选择原论文中第三种方法，即对四类因子进行平均分为高低两组，即2*2*2*2=16组分组结果
    7.股票池选择为中证800作为样本，市场组合收益率为中证800指数


```python
import numpy as np
import pandas as pd 
def get_16parts(change_pool_time):  #更新因子生成的投资组合函数，change——pool——time为更好组合时间，取每年的6月7日
    cpt=change_pool_time
    C=DataAPI.MktEqudGet(ticker='002053',beginDate=cpt,endDate=str(int(cpt)+20),field=u"ticker,tradeDate")
    cpt=filter(lambda x:x.isdigit(),C.iat[len(C)-1,1])
    universe=set_universe('000906.ZICN',date=cpt) #股票池定为中证800
    size=DataAPI.MktEqudGet(tradeDate=cpt,secID=universe,field=u"ticker,marketValue").dropna()  #选取市值
    size50=np.percentile(size['marketValue'],50)
    size0=size[size['marketValue']<=size50]['ticker'].tolist()  #之后一样，市值分组为0代表数值高一组，1代表低一组
    size1=size[size['marketValue']>size50]['ticker'].tolist()
    #市净率数据
    BP=DataAPI.MktStockFactorsOneDayGet(tradeDate=cpt,secID=universe,field=u"ticker,PB").dropna()
    BP=BP[BP>0].dropna()                                                  #去掉PB值为负的股票
    BP[['PB']]=1/BP[['PB']]
    BP50=np.percentile(BP['PB'],50)
    BP0=BP[BP['PB']<=BP50]['ticker'].tolist()
    BP1=BP[BP['PB']>BP50]['ticker'].tolist()
    #净资产收益率数据
    ROE=DataAPI.MktStockFactorsOneDayGet(tradeDate=cpt,secID=universe,field=u"ticker,ROE").dropna()
    ROE50=np.percentile(ROE['ROE'],50)
    ROE0=ROE[ROE['ROE']<=ROE50]['ticker'].tolist()
    ROE1=ROE[ROE['ROE']>ROE50]['ticker'].tolist()
    #投资，即总资产增长数据
    cg=DataAPI.MktStockFactorsOneDayGet(tradeDate=cpt,secID=universe,field=u"ticker,TotalAssetGrowRate").dropna()
    cg50=np.percentile(cg['TotalAssetGrowRate'],50)
    cg0=cg[cg['TotalAssetGrowRate']<=cg50]['ticker'].tolist()
    cg1=cg[cg['TotalAssetGrowRate']>cg50]['ticker'].tolist()
    #按顺序市值、BP、净资产、投资分为16类组合，其中记录为，例如，高市值，低BP，低roe，高cg为part0110
    part0000=list(set(size0).intersection(set(BP0),set(ROE0),set(cg0)))
    part0001=list(set(size0).intersection(set(BP0),set(ROE0),set(cg1)))
    part0011=list(set(size0).intersection(set(BP0),set(ROE1),set(cg1)))
    part0111=list(set(size0).intersection(set(BP1),set(ROE1),set(cg1)))
    part1111=list(set(size1).intersection(set(BP1),set(ROE1),set(cg1)))
    part0010=list(set(size0).intersection(set(BP0),set(ROE1),set(cg0)))
    part0110=list(set(size0).intersection(set(BP1),set(ROE1),set(cg0)))
    part1110=list(set(size1).intersection(set(BP1),set(ROE1),set(cg0)))
    part0100=list(set(size0).intersection(set(BP1),set(ROE0),set(cg0)))
    part1100=list(set(size1).intersection(set(BP1),set(ROE0),set(cg0)))
    part1000=list(set(size1).intersection(set(BP0),set(ROE0),set(cg0)))
    part1010=list(set(size1).intersection(set(BP0),set(ROE1),set(cg0)))
    part1001=list(set(size1).intersection(set(BP0),set(ROE0),set(cg1)))
    part0101=list(set(size0).intersection(set(BP1),set(ROE0),set(cg1)))
    part1011=list(set(size1).intersection(set(BP0),set(ROE1),set(cg1)))
    part1101=list(set(size1).intersection(set(BP1),set(ROE0),set(cg1)))
    return part0000,part0001,part0010,part0011,part0100,part0101,part0110,part0111,part1000,part1001,part1010,part1011,part1100,part1101,part1110,part1111
```

构建完投资组合，我们需要得到投资组合的月度收益率情况，构建函数组合收益率获取，由马科维兹的CAMP有在市值加权的收益率为组合的最小风险收益率，所以我们组合收益率为成分股市值加权形成的收益率。(因为组合一年调仓一次，所以函数针对年，计算一年内所以月度组合收益率)


```python
def yield_monthly(x,Year):
    from CAL.PyCAL import *
    data=DataAPI.TradeCalGet(exchangeCD=u"XSHG",beginDate=str(Year*10000+607),endDate=str((Year+1)*10000+707),field=['calendarDate','isMonthEnd'])
    data = data[data['isMonthEnd'] == 1]
    date= map(lambda x: x[0:4]+x[5:7]+x[8:10], data['calendarDate'].values.tolist())
     #调用投资组合x每只股票每个月末的市值，收盘价用来计算收盘价
    returnMonthly=np.zeros(12)
    for i in range(12):
        inf1=DataAPI.MktEqudAdjGet(tradeDate=date[i],ticker=x,field=u"ticker,closePrice").set_index('ticker')
        inf2=DataAPI.MktEqudAdjGet(tradeDate=date[i+1],ticker=x,field=u"ticker,marketValue,closePrice"\
        ).set_index('ticker')
        Return=pd.concat([inf2,inf1],axis=1)
        Return.columns=['Weight','Return','WReturn']                   #计算每只股票收益率和市值加权的权重以及两者的乘积
        Return['Weight']=Return['Weight']/Return['Weight'].sum()
        Return['Return']=Return['Return']/Return['WReturn']-1
        Return['WReturn']=Return['Weight']*Return['Return']
        returnMonthly[i]=Return['WReturn'].sum()
    return returnMonthly

```

因子数据的生成
市值因子（SMB）： average(part0xxx)-average(part1xxx)
bp因子(HML):average(partx0xx)-average(partx0xx)
盈利因子（RMW）：average(partxx0x)-average(partxx1x)
投资因子(CMA)：average(part0xx0)-average(part1xx1)
因子的数据是月度，所以因子的长度应该为年数*12以2008年到2018年10年跨度数据，即有120个数据，四组因子

```python
SMB=[]
HML=[]
RMW=[]
CMA=[]
r_groups=pd.DataFrame()  #用于存储每个组合的月收益率序列，方便我们之后查看
r_groups['p0000']=np.zeros(120)
r_groups['p0001']=np.zeros(120)
r_groups['p0010']=np.zeros(120)
r_groups['p0011']=np.zeros(120)
r_groups['p0100']=np.zeros(120)
r_groups['p0101']=np.zeros(120)
r_groups['p0110']=np.zeros(120)
r_groups['p0111']=np.zeros(120)
r_groups['p1000']=np.zeros(120)
r_groups['p1001']=np.zeros(120)
r_groups['p1010']=np.zeros(120)
r_groups['p1011']=np.zeros(120)
r_groups['p1100']=np.zeros(120)
r_groups['p1101']=np.zeros(120)
r_groups['p1110']=np.zeros(120)
r_groups['p1111']=np.zeros(120)
for Year in range(2008,2018):
    p0000,p0001,p0010,p0011,p0100,p0101,p0110,p0111,p1000,p1001,p1010,p1011,p1100,p1101,p1110,p1111=\
    get_16parts(str(Year*10000+607))
    r_kind=range(16)
    labels=[p0000,p0001,p0010,p0011,p0100,p0101,p0110,p0111,p1000,p1001,p1010,p1011,p1100,p1101,p1110,p1111]
    for Kind in range(16):
        r_kind[Kind]=yield_monthly(labels[Kind],Year)
        r_groups.iloc[(Year-2008)*12:(Year-2007)*12,[Kind]]=r_kind[Kind].reshape(12,1)
    SMBr=(r_kind[0]+r_kind[1]+r_kind[2]+r_kind[3]+r_kind[4]+r_kind[5]+r_kind[6]+r_kind[7])-\
    (r_kind[8]+r_kind[9]+r_kind[10]+r_kind[11]+r_kind[12]+r_kind[13]+r_kind[14]+r_kind[15])        
    HMLr=(r_kind[0]+r_kind[1]+r_kind[2]+r_kind[3]+r_kind[8]+r_kind[9]+r_kind[10]+r_kind[11])-\
    (r_kind[4]+r_kind[5]+r_kind[6]+r_kind[7]+r_kind[12]+r_kind[13]+r_kind[14]+r_kind[15])
    RMWr=(r_kind[0]+r_kind[1]+r_kind[4]+r_kind[5]+r_kind[8]+r_kind[9]+r_kind[12]+r_kind[13])-\
    (r_kind[2]+r_kind[3]+r_kind[6]+r_kind[7]+r_kind[10]+r_kind[11]+r_kind[14]+r_kind[15])
    CMAr=(r_kind[0]+r_kind[2]+r_kind[4]+r_kind[6]+r_kind[8]+r_kind[10]+r_kind[12]+r_kind[14])-\
    (r_kind[1]+r_kind[3]+r_kind[5]+r_kind[7]+r_kind[9]+r_kind[11]+r_kind[13]+r_kind[15])
    for i1 in range(12):
        SMBr[i1]=SMBr[i1]/8
    for i2 in range(12):
        HMLr[i2]=HMLr[i2]/8
    for i3 in range(12):
        RMWr[i3]=RMWr[i3]/8
    for i4 in range(12):
        CMAr[i4]=CMAr[i4]/8
    SMB +=SMBr.tolist()
    HML +=HMLr.tolist()
    RMW +=RMWr.tolist()
    CMA +=CMAr.tolist()
SMB=np.array(SMB)
HML=np.array(HML)
RMW=np.array(RMW)
CMA=np.array(CMA)
```

回归：
  已经完成因子数据的生成，进行回归，验证五因子模型的有效性。 这里共有共有四个因子，进行分组，若我们采取惯例分为5组，进行回归，即会625次回归，考虑增加变量，导致拟合效果自然的提升，我们可以适当减少分组，故仅进行3次分组，即只进行81次回归。先构建81个组合，同时记录他们的月度收益率，然后对应之前计算出的因子，进行回归。
回归方程为：
    R(t)−Rf(t)=a+b(RM(t)−Rf(t))+sSMB(t)+hHML(t)+RRMW(t）+CCMA(t)+e(t)
    回归才用最小二乘，有截距项回归

```python
def get_64parts(change_pool_time):  #更新因子生成的投资组合函数，change——pool——time为更好组合时间，取每年的6月7日
    cpt=change_pool_time
    C=DataAPI.MktEqudGet(ticker='002053',beginDate=cpt,endDate=str(int(cpt)+20),field=u"ticker,tradeDate")
    cpt=filter(lambda x:x.isdigit(),C.iat[len(C)-1,1])
    universe=set_universe('000906.ZICN',date=cpt) #股票池定为中证800
    size=DataAPI.MktEqudGet(tradeDate=cpt,secID=universe,field=u"ticker,marketValue").dropna()  #选取市值
    #市净率数据
    BP=DataAPI.MktStockFactorsOneDayGet(tradeDate=cpt,secID=universe,field=u"ticker,PB").dropna()
    BP=BP[BP>0].dropna()                                                  #去掉PB值为负的股票
    BP[['PB']]=1/BP[['PB']]
    #净资产收益率数据
    ROE=DataAPI.MktStockFactorsOneDayGet(tradeDate=cpt,secID=universe,field=u"ticker,ROE").dropna()
    ROE50=np.percentile(ROE['ROE'],50)
    ROE0=ROE[ROE['ROE']<=ROE50]['ticker'].tolist()
    ROE1=ROE[ROE['ROE']>ROE50]['ticker'].tolist()
    #投资，即总资产增长数据
    cg=DataAPI.MktStockFactorsOneDayGet(tradeDate=cpt,secID=universe,field=u"ticker,TotalAssetGrowRate").dropna()
    cg50=np.percentile(cg['TotalAssetGrowRate'],50)
    cg0=cg[cg['TotalAssetGrowRate']<=cg50]['ticker'].tolist()
    cg1=cg[cg['TotalAssetGrowRate']>cg50]['ticker'].tolist()
    #构建size和bp
    sizeq=range(5)
    BPq=range(5)
    size_5=pd.Series()
    BP_5=pd.Series()
    ROE_5=pd.Series()
    cg_5=pd.Series()
    for i in range(4):
        sizeq[i+1]=np.percentile(size['marketValue'],(i+1)*25)                           #算出市值大小的20%,40%,60%,80%分位值
        BPq[i+1]=np.percentile(BP['PB'],(i+1)*25)
        D=size[(size['marketValue']>sizeq[i]) & (size['marketValue']<=sizeq[i+1])]['ticker'].tolist()     #取市值处于相应分位值区间的股票
        size_5=pd.concat([size_5,pd.Series(D)],axis=1)                                 #存于dataframe里
        E=BP[(BP['PB']>BPq[i]) & (BP['PB']<=BPq[i+1])]['ticker'].tolist()
        BP_5=pd.concat([BP_5,pd.Series(E)],axis=1)
    size_5.columns=[5,0,1,2,3] #重命名列名
    BP_5.columns=[5,0,1,2,3]
    #构造ROE，cg
    ROE_5=pd.concat([pd.Series(ROE1),pd.Series(ROE0)],axis=1)
    cg_5=pd.concat([pd.Series(cg1),pd.Series(cg0)],axis=1)
    ROE_5.columns=range(2)
    cg_5.columns=range(2)
    BP_5=BP_5.drop(labels=[5],axis=1)
    size_5=size_5.drop(labels=[5],axis=1)
    Group625=pd.Series([]) #用于存著交叉取交集后的625个股票组合
    for a1 in range(4):
        for a2 in range(4):
            for a3 in range(2):
                for a4 in range(2):
                    s1=size_5[a1].dropna().tolist()
                    s2=BP_5[a2].dropna().tolist()
                    s3=ROE_5[a3].dropna().tolist()
                    s4=cg_5[a4].dropna().tolist()
                    stocks=pd.Series(list(set(s1).intersection(set(s2),set(s3),set(s4))))
                    Group625=pd.concat([Group625,stocks],axis=1)
    Group625.columns=range(65)
    Group625=Group625.drop(axis=1,labels=[0])
    Group625.columns=range(64)
    return Group625
```

```python
EReturn=np.zeros((64,12*10)) #用于存储25个组合的超额收益序列
for i in range(64):
    a=[]
    print(i)
    for Year in range(2008,2018):
        Group64=get_64parts(str(Year*10000+607))   #每年进行分组
        a=a+(yield_monthly(Group64[i].dropna().tolist(),Year)).tolist()   #收益率转化为list，方便每年相加
    EReturn[i]=np.array(a)-RfMonthly
```

```python
#CAMP MODEL
from sklearn import linear_model
a1=np.zeros(64)   #a项
b1=np.zeros(64)   #市场因子项系数
e1=np.zeros((64,120))   #残差项
R2_1=np.zeros(64)   #R2相关系数平方
tb1=np.zeros(64)
ta1=np.zeros(64)
import statsmodels.api as sm
from statsmodels.sandbox.regression.predstd import wls_prediction_std
ap1=np.zeros(64)  #a显著性检验的P值，下面类同
bp1=np.zeros(64)
for i in range(64):
    X=np.zeros((1,120))
    X[0]=MF
    X=X.T
    X = sm.add_constant(X,has_constant='skip')
    y=EReturn[i]
    model = sm.OLS(y, X)
    results = model.fit()
    a1[i] = results.params[0]
    b1[i] = results.params[1]
    ap1[i]=results.pvalues[0]
    bp1[i]=results.pvalues[1]
    R2_1[i] = results.rsquared
    e1[i] = results.resid
    tb1[i] = results.tvalues[1]
    ta1[i] = results.tvalues[0]
```

```python
#模型2，三因子模型，选取MF,SMB,HML三个因子
import numpy as np
from sklearn import linear_model
a3=np.zeros(64)   #a项
b3=np.zeros(64)   #市场因子项系数
s3=np.zeros(64)   #规模因子项系数
h3=np.zeros(64)   #账面价值比项系数
e3=np.zeros(64)   #残差项
R2_2=np.zeros(64)   #R2相关系数平方
for i in range(64):
    x=np.zeros((3,120))
    x[0]=MF
    x[1]=SMB
    x[2]=HML
    y=EReturn[i]
    x=np.mat(x).T
    y=np.mat(y).T
    regr = linear_model.LinearRegression()
    regr.fit(x,y)
    b3[i]=regr.coef_[0][0]
    s3[i]=regr.coef_[0][1]
    h3[i]=regr.coef_[0][2]
    a3[i]=regr.intercept_[0]
    e3[i]=regr.residues_
    R2_2[i]=regr.score(x,y)
```
