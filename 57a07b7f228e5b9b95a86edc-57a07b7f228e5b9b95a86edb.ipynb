```python
#CCI低于-110且回升至-100以上是买入，CCI高于110且回落100以下卖出

#指标2  ：WR
def get_WR(secID = '000001.ZICN',date = '',N = 10):
    """获取secID的日期为date那天的WR值，参数默认为10，返回WR为单个float
    """
    tradedates = pre_n_tradedays(date, n=N,with_ = False)
    data = DataAPI.MktIdxdGet(indexID=secID,beginDate = tradedates[-1],endDate=date,field=u"tradeDate,highestIndex,lowestIndex,closeIndex",pandas="1")
    HIGH = data['highestIndex'].max()
    LOW = data['lowestIndex'].min()
    CLOSE = data['closeIndex'][data.index[-1]]
    WR = 100 * (HIGH - CLOSE)/ (HIGH - LOW) 
    return WR

#指标值处理函数：
def high_to_low(data_list,height,depth,search_window = 10):
    """判断在search_window个时间单位是内否从高值到低值，输入列表，返回bool
    """
    if max(data_list[-search_window:]) > height and data_list[-1] < depth:
        return True
    else:
        return False
def low_to_high(data_list,depth,height,search_window = 3):
    """判断在search_window个时间单位是否从低值到高值，输入列表，返回bool
    """
    if min(data_list[-search_window:]) < depth and data_list[-1] > height:
        return True
    else:
        return False
```
