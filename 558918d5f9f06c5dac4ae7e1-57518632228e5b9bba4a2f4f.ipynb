```python
import pandas as pd

# 获取沪深300成分股
hs300 = set_universe('HS300')

#沪指在6月12号达到本轮牛市以来最高值,随后一路下跌.
#从6月12号开始的数据
df = DataAPI.MktEqudAdjGet(hs300, beginDate=start_date)

# 预览最后5条数据
df.tail()
```

```python
# 按ticker名分类, 计算6月12号以来的涨跌幅.
percent_df = pd.DataFrame(df.groupby(['ticker', 'secShortName'])['closePrice'].last()/df.groupby('ticker')['closePrice'].first()-1)

# 按跌幅排序
percent_df = percent_df.sort('closePrice', ascending=1)

# 预览前30条
percent_df.head(30)
```
