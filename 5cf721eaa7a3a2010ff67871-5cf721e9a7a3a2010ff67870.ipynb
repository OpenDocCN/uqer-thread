```python
#10秒读懂分析逻辑：大盘指数与个股 图形比较，短期看个股跟随指数，长期看如何实现量变，炒妖股与价值投资的难易，指数当做市场平均线来看，做好一直股票就可以！
#第一个图：指数和2只股票，第二个图，股票K险图
#沟通VX：h1149351843
import numpy as np
import pandas as pd
import talib
import datetime 
import matplotlib.dates as mdates
import matplotlib.pyplot as plt
from matplotlib.finance import candlestick_ohlc  
from dateutil.parser import parse
#=========================================================================================================   input
#['000300','000905','000016','000001','399001']  #大盘 ['000300'，沪深300,'000905'中证500,'000016',上证50,'000001','399001']
#['000922','000827','000991','000990','000993','000992']#指数
#'000016',上证50,'399101'，中小板,'399006',创业板,000905,中证500,000300，沪深300，'000922',中证红利，'399330',深圳100，
#'000993'，全指信息，000992，全指金融，000990，全指消费，000991，全指医药，000827，全指环保
#--------------------------------------------------------------------------------------- 先时间参数 后ticker参数
td_begin =u"20100101" 
plot_td_sec = '2016-05-01'

#sym_ = ['000300',000016','000001']
ticke_idx = u"000001"   
p_sym_sec =['600519','000025'] #贵州茅台 特立A
plot_ticker_sec = '600519'
#=========================================================================================================  cal
def defPlot_idx_sec(ticke_idx,p_sym_sec,td_begin):
    pd_idx = pd.DataFrame()              
    pd_idx = DataAPI.MktIdxdGet(ticker=ticke_idx,beginDate=td_begin,endDate=u"",field=u"closeIndex",pandas="1") 
    td_sec = DataAPI.MktEqudGet(ticker=u"601998",beginDate=td_begin,endDate=u"",isOpen="",field=u"tradeDate",pandas="1")
    pd_idx.index = pd.to_datetime(td_sec.tradeDate)
    #print(pd_idx) 

    pd_sec = pd.DataFrame()
    for sym in p_sym_sec:                    
        pd_sec[sym]=DataAPI.MktEqudAdjGet(ticker=sym,isOpen="",beginDate=td_begin,endDate=u"",field=u"closePrice",pandas="1")
    #print(df4.head()) 
    td_sec = DataAPI.MktEqudGet(ticker=u"601998",beginDate=td_begin,endDate=u"",isOpen="",field=u"tradeDate",pandas="1")
    pd_sec.index = pd.to_datetime(td_sec.tradeDate)
    #print(pd_sec)
    #==========================================================================================================  plot
    fig = plt.figure()    # 多条线合一
    ax = fig.add_axes([0.1,0.1,2.4,1.6])
    for sym in p_sym_sec:   
        ax.plot(pd_sec[sym]/pd_sec[sym].ix[0]*100,label = sym)
    ax.plot(pd_idx/pd_idx.ix[0]*100,'k',label = ticke_idx)
    ax.legend(loc='best')
    
defPlot_idx_sec(ticke_idx,p_sym_sec,td_begin)
#-----------------------------------------------------------------------------------------------------------
pd_sec = DataAPI.EquGet(ticker=p_sym_sec,listStatusCD=u"",field=u"ticker,secShortName,ListSector,primeOperating",pandas="1")
print(pd_sec)
#------------------------------------------------------------------------------------------------------------
def get_candlestick_plot(ticker,start_date,end_date='-'):
    start_date = parse(start_date).strftime('%Y%m%d')
    end_date = parse(end_date).strftime('%Y%m%d')
    df_history = DataAPI.MktEqudAdjGet(secID=u"",ticker=ticker,beginDate=start_date,endDate=end_date,isOpen="",
                   field=u"tradeDate,openPrice,highestPrice,lowestPrice,closePrice",pandas="1")
    df_history = df_history[df_history['openPrice'] != 0]  ## 去除停牌的行情
    df_history['tradeDate'] = df_history['tradeDate'].map(lambda x:mdates.date2num(datetime.datetime.strptime(x,'%Y-%m-%d')))
    df_history['ii'] = range(len(df_history))
    date = df_history['ii'].tolist()
    ohlc = df_history['ii'].map(lambda x:tuple(df_history.iloc[x][['tradeDate','openPrice','highestPrice','lowestPrice','closePrice']])).tolist()
    weekday_ohlc = [tuple([i]+list(item[1:])) for i,item in enumerate(ohlc)]
    
    fig = plt.figure(figsize=(18,5))   #figsize=(len(weekday_ohlc)*10.0/46,7)
    ax = plt.subplot(1,1,1)
    candlestick_ohlc(ax, weekday_ohlc, width=0.6, colorup='r', colordown='g')
    ax.set_xticks(range(0,len(weekday_ohlc),5))### 每五天标一个日期
    ax.set_xticklabels([mdates.num2date(ohlc[index][0]).strftime('%Y-%m-%d') for index in ax.get_xticks()])####
    ax.legend([ticker],frameon=False)
    plt.xlim(0,len(weekday_ohlc)-1) #设置一下x轴的范围
    plt.setp(plt.gca().get_xticklabels(), rotation=45, horizontalalignment='right') #将x轴的label转一下，好看一点
    plt.show()
#-----------------------------------------------------------------------------------------------------------------    
get_candlestick_plot(plot_ticker_sec,plot_td_sec)   #不写end_date默认今天
```
