```python
import scipy.fftpack as fftpack
import numpy as np
import pandas as pd
import datetime
from CAL.PyCAL import *
import matplotlib.pyplot as pl
```

```python
a = DataAPI.MktIdxdGet(tradeDate=u"",indexID=u"",ticker=u"000002",beginDate=u"20150101",endDate=u"20160101",field=u"",pandas="1")
```

巴拉巴拉各种定义！这个不重要！数据用的是上证指数最近一年数据！

```python
a['diff'] = a['closeIndex'].diff()
a['diff1'] = a['diff'].diff()
a = a.dropna()
a['hilbert'] = fftpack.hilbert(a['diff1'])
a = a.dropna()
```

引入hilbert空间，刚好！！！pythonl里面刚刚好有这个函数！就直接拿来用了！至于原理就是把周期量化！这里有一个！就是去除趋势项！意思就是，时间序列的数据有噪声，那么要去噪处理，还有一个就是长期趋势项，接着就是短期的波动项了。我们要做的就是量化短期的波动项！来从非周期中强行找周期！因为我记得之前上课有同学用ARIMA模型研究过大盘指数，说的是二阶差分平稳！所以，我就大胆用了二阶差分！（其实是一阶差分分类效果不好，所以用的二阶，刚好又想到之前同学做的真的是二阶差分的ARIMA模型，所以才这么说。）

```python
def zd(x):
    if x>0:
        a = 1
    else:
        a = 0
    return a
```

涨跌定义，涨是1，跌势0。所以，均值就是涨的概率。

```python
def hb(x):
    if x>0:
        a = 0
    else:
        a = 1
    return a
```

hilbert函数定义，看涨是1看跌是0。

```python
a['zd'] = a['CHG'].apply(zd)
a['hb'] = a['hilbert'].apply(hb)
```

```python
c = a['zd'].groupby([a['hb']]).mean()
```
