```python
############################################################################################

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
import sys
import time
import calendar
from dateutil.relativedelta import relativedelta
import datetime
import math
import traceback
import matplotlib as mpl
mpl.rcParams['font.sans-serif']=["SimHei"]#设置中文
mpl.rcParams["axes.unicode_minus"]=False
mpl.rcParams['pdf.fonttype'] = 42
mpl.rcParams['ps.fonttype'] = 42
import re


############################################################################################

def get_salary_value(salary_range):
    '''
    :param salary_range: 招聘网站上爬下来的薪资范围
    :return: [年薪最小值，年薪最大值], 如果无法判断或者范围异常，则返回['invalid','invalid']
    '''
    def get_start_end(salary_range_str):
        salary_range = salary_range_str.replace("/年", "").replace("每年", "").replace("年", "").replace("元", "")#将年去除
        salary_range = salary_range.split("/")[0]
        start_r, end_r = salary_range.split("-")
        # 先判断起始是否有单位
        if 'k' in start_r or 'K' in start_r or '千' in start_r:
            start_mult = 1000
            start_num = str(start_r).replace(" ", "").replace('k', "").replace("K", "").replace("千", "")
            start_num = float(start_num)
        elif 'w' in start_r or 'W' in start_r or '万' in start_r:
            start_mult = 10000
            start_num = str(start_r).replace(" ", "").replace('w', "").replace("W", "").replace("万", "")
            start_num = float(start_num)
        else:
            start_mult = None  # 无单位则跟随后面的
            start_num = str(start_r).replace(" ", "").replace('w', "").replace("W", "").replace("万", "")
            start_num = float(start_num)

        # 确定结束的单位
        end_r = str(end_r).replace(" ", "").replace("·", "*")
        end_r = end_r.split("*")[0]

        # 先判断起始是否有单位
        if 'k' in end_r or 'K' in end_r or '千' in end_r:
            end_mult = 1000
            end_num = str(end_r).replace(" ", "").replace('k', "").replace("K", "").replace("千", "")
            end_num = float(end_num)
        elif 'w' in end_r or 'W' in end_r or '万' in end_r:
            end_mult = 10000
            end_num = str(end_r).replace(" ", "").replace('w', "").replace("W", "").replace("万", "")
            end_num = float(end_num)
        else:
            end_mult = None  # 无单位则跟随后面的
            end_num = str(end_r).replace(" ", "").replace('w', "").replace("W", "").replace("万", "")
            end_num = float(end_num)
         
        if start_mult is None:
            if start_num>=1000:
                start_mult = 1
            else:
                if end_mult is None:
                    if end_num < 2000:
                        return 'invalid', 'invalid'  # 最大收入小于2000，直接置为无效
                    else:
                        start_mult = end_mult = 1
                else:
                    start_mult = end_mult    

        start_salary = start_num * start_mult
        end_salary = end_num * end_mult
        
        if "年" in salary_range_str:#判断是否有年字样
            start_salary_year =  start_salary
            end_salary_year = end_salary
        else:
            # 所有数据没有年字样，默认为月薪
            start_salary_year = 12 * start_salary
            end_salary_year = 12 * end_salary

        return start_salary,end_salary,start_salary_year,end_salary_year

    try:
        
        if '-' not in salary_range:#salary_range一列中没有-如何处理
            if "面议" in salary_range or "天" in salary_range or "小时" in salary_range:
                return 'invalid', 'invalid'
            else:
                pattern = re.compile('\d+\.?\d?')
                salary_num = int(re.findall(pattern,salary_range)[0])
                if 'k' in salary_range or 'K' in salary_range or '千' in salary_range:
                    if "月" in salary_range:
                        start_salary_year = end_salary_year = salary_num *1000*12
                    elif "年" in salary_range:
                        start_salary_year = end_salary_year = salary_num *1000
                    else:
                        if (salary_num *1000*12)>=5000000:
                            start_salary_year = end_salary_year = salary_num *1000
                        else:
                            start_salary_year = end_salary_year = salary_num *1000*12
                        
                elif 'w' in salary_range or 'W' in salary_range or '万' in salary_range:
                    if "月" in salary_range:
                        start_salary_year = end_salary_year = salary_num *10000*12
                    elif "年" in salary_range:
                        start_salary_year = end_salary_year = salary_num *10000
                    else:
                        if salary_num>5:
                            start_salary_year = end_salary_year = salary_num *10000
                        else:
                            start_salary_year = end_salary_year = salary_num *10000*12  
                else:
                    if "月" in salary_range:
                        start_salary_year = end_salary_year = salary_num *12
                    else:
                        start_salary_year = end_salary_year = salary_num 
                return [start_salary_year, end_salary_year]
                    
        else:# salary_range一列中有-如何处理         
            if "月" in salary_range:#如果没有年，统一按照月薪计算处理
                salary_range = salary_range.replace("/月", "").replace("每月", "").replace("月", "").replace("元", "")
                start_salary,end_salary,start_salary_year,end_salary_year=get_start_end(salary_range)
                # 如果年薪异常高，则恢复为月薪或者置为无效
                if start_salary_year >= 4000000:
                    start_salary_year = 'invalid'
                if end_salary_year >= 5000000:
                        end_salary_year = 'invalid'
                # 如果年薪异常低，则置为无效
                if start_salary_year <= 10000:
                    start_salary_year = 'invalid'
                if end_salary_year <= 10000:
                    end_salary_year = 'invalid'
                return [start_salary_year, end_salary_year]

            elif "年" in salary_range:#如果有年，按照年薪处理
                start_salary,end_salary,start_salary_year,end_salary_year=get_start_end(salary_range)
                
                return [start_salary_year, end_salary_year]
            
            else:#都没有单位字样的时候按照月薪处理，但是需要筛选一些条件
                salary_range = salary_range.replace("元", "")
                start_salary,end_salary,start_salary_year,end_salary_year=get_start_end(salary_range)
                # 如果年薪异常高，则恢复为月薪或者置为无效
                mean_salary_month = 0.5*(start_salary+end_salary)
                mean_salary_year = 0.5*(start_salary_year+end_salary_year)

                if 'w' in salary_range or 'W' in salary_range or '万' in salary_range:
                    if mean_salary_month <= 50000:
                        start_salary_year=start_salary_year
                        end_salary_year=end_salary_year
                    
                    elif mean_salary_year > 5000000:
                        start_salary_year='invalid'
                        end_salary_year = 'invalid'
                    else:
                        start_salary_year = start_salary
                        end_salary_year = end_salary
                else:
                    if mean_salary_year > 5000000:
                        start_salary_year='invalid'
                        end_salary_year = 'invalid'
                    else:
                        start_salary_year=start_salary_year
                        end_salary_year=end_salary_year
                    
            # 如果年薪异常低，则置为无效
            if start_salary_year <= 10000:
                start_salary_year = 'invalid'
            if end_salary_year <= 10000:
                end_salary_year = 'invalid'
            return [start_salary_year, end_salary_year]
    
    except:
        return ['invalid', 'invalid']


# 针对meta表的数据，每一个记录，增加平均薪资
def add_average_salary_feature(data_df):
    '''
    :param data_df: meta表的数据，列至少包括 salary_range
    :return:  data_df，新增 'average_salary','salary_start','salary_end'列  （年薪,float类型）
    '''
    data_df['salary_list'] = data_df.apply(lambda x: get_salary_value(x['salary_range']), axis=1)
    data_df['salary_start']=data_df['salary_list'].apply(lambda x:x[0]).replace("invalid", np.nan)
    data_df['salary_end']=data_df['salary_list'].apply(lambda x:x[1]).replace("invalid", np.nan)
    
    # 计算平均年薪
    data_df['average_salary'] = data_df[['salary_start', 'salary_end']].mean(axis=1, skipna=True)
    return data_df


# 增加工作经验的标签
def add_experience_feature(data_df):
    '''
    :param data_df: meta表的数据，至少包含  working_experience列
    :return: 增加列 'work_expe', 工作经验是否应届生，工作经验是否经验不限, ...
    '''
    regex = r"\d+"
    data_df['work_expe'] = data_df['working_experience'].str.findall(regex).apply(
        lambda x: '经验不限' if x is None else (
            min([int(i) for i in x]) if len(x) != 0 else ('应届毕业生' if x == '应届毕业生' else '经验不限')))
    data_df['work_expe']=np.where(data_df['working_experience'].str.contains('应届'),'应届毕业生',data_df['work_expe'])

    data_df['工作经验是否应届生'] = data_df['work_expe'].apply(lambda x: 1 if x == '应届毕业生' else 0)
    data_df['工作经验是否经验不限'] = data_df['work_expe'].apply(lambda x: 1 if x == '经验不限' else 0)
    data_df['工作经验是否0-3年'] = data_df['work_expe'].apply(
        lambda x: 0 if type(x) == str else (1 if (x >= 0 and x < 3) else 0))
    data_df['工作经验是否3-5年'] = data_df['work_expe'].apply(
        lambda x: 0 if type(x) == str else (1 if (x >= 3 and x < 5) else 0))
    data_df['工作经验是否5年以上'] = data_df['work_expe'].apply(lambda x: 0 if type(x) == str else (1 if (x >= 5) else 0))
    return data_df

def get_education(x):
    if x is None:
        return '不限'
    if '大专' in x:
        return '其他学历'
    elif '本科' in x:
        return '本科'
    elif '硕士' in x:
        return '硕士'
    elif '博士' in x:
        return '博士'
    elif '不限' in x:
        return "不限"
    elif '中技' in x:
        return '其他学历'
    elif '高中' in x:
        return '其他学历'
    else:
        return '其他学历'#将初中以及其他学历全部归为初中方便统计
    
def trans_exp_tag(exp_str):
    if exp_str=='应届毕业生':
        tag = '应届生'
    elif exp_str=='经验不限':
        tag = '不限'
    elif (type(exp_str)!=str) and (exp_str>=0) and (exp_str<3):
        tag = '0-3年'
    elif (type(exp_str)!=str) and (exp_str>=3) and (exp_str<5):
        tag = '3-5年'
    elif (type(exp_str)!=str) and (exp_str>=5):
        tag = '5年以上'
    else:
        tag = '不限'
    return tag

def get_quarter(month_str):
    '''
    month_str:月度日期
    '''
    if month_str[-2:] in ['01','02','03']:
        quarter_str = 'Q1'
    elif month_str[-2:] in ['04','05','06']:
        quarter_str = 'Q2'
    elif month_str[-2:] in ['07','08','09']:
        quarter_str = 'Q3'
    else:
        quarter_str = 'Q4'
    return quarter_str

def transform_quater_df(data_df):
    #将周度以及月度转换成季度的数据进行分析
    '''
    data_df：需要转换的招聘数据,要求包含RECRUITMENT_SALARY、RECRUITMENT_NUM和PERIOD_DATE
    '''
    data_df['all_sum_salary'] = data_df['RECRUITMENT_SALARY']*data_df['RECRUITMENT_NUM']
    data_df['month'] = data_df['PERIOD_DATE'].apply(lambda x:str(x)[:7])
    data_df['quarter'] = data_df['month'].apply(lambda x:str(x)[:4]+get_quarter(x))
    
    return data_df
```
