# PE因子选股策略20201106

市盈率PE(price earning ratio)=估价price/每股盈余earnings per share，是常用的衡量估价是否合理的指标。
市盈率一般保持在20到30之间，市盈率小说明股价低，有上涨空间，风险小，值得购买。但是PE因行业差距较大，不能一概而论。
市盈率高说明市场对该股预期极高，一般需要有足够的理由支持，一旦公司的盈利能力不及市场期望，股价往往会下跌。因此市盈率高的股票风险大，往往需谨慎。
因此我们的基本策略是购买市盈率低的股票，每月初调仓交易。
为了使因子和收益率成正相关，我们以市盈率的倒数(成为EP)作为选股的单因子，对此因子以及投资权重逐步做调整，进行回测，观察去年十年市场的表现(从20100602到20170602)。

(注:数据来源是优矿的API接口，自己的数据都可以没差的)

1.纯EP因子选股
这种方法最简单粗暴，即每此交易都选择EP大的股票，每月初调仓，根据上一月最后一个交易日的EP因子选取表现最好的20支股票，按照因子值分配权重进行投资。

```python
#导入所需包裹以便后面使用
import pandas as pd
import numpy as np
import statsmodels.api as sm
from cvxopt import matrix,solvers
import matplotlib.pyplot as plt
```

```python
#获取从20100602到20170602的每月末交易日历 
#(为什么要选取每月末？因为对于每个月末，可以获取每只股的实际收益率进而获得portfolio的实际收益率，并且因子值的计算时间也是月末)
calendar=DataAPI.TradeCalGet(exchangeCD=u"XSHG,XSHE",beginDate=u"20100602",endDate=u"20180602",field=u"calendarDate,isMonthEnd",pandas="1")  #使用API获取交易日历
calendar=calendar[calendar['isMonthEnd']==1]['calendarDate'].tolist()  #选取月末，转化为list
calendar[:5] #取前五个观察list是否正确
```

```python
#获取每个交易日每只股的EP值 (此处使用API接口获取数据，获取数据的过程不必细说)
factor=pd.DataFrame()
for td in calendar: 
    tmptfac=DataAPI.MktStockFactorsOneDayGet(tradeDate=td,secID=u"",ticker=u"",field=u"secID,tradeDate,PE",pandas="1")
    tmptfac['EP']=1/tmptfac['PE']
    tmptfac.drop('PE',axis=1,inplace=True)
    factor=factor.append(tmptfac)
```

```python
#获取了因子,组成分别是股票编码，时间和因子值
factor.head(10)
```

```python
#因为每月初的调仓是基于上个月末的因子，所以对于一个月末的收益率，因子其实是对应上个月末的，切记不能偷窥未来数据！
#比如说2010-07-30的收益率，选取它出来的因子是2010-06-30的因子。因此，对于每只股票，都将因子滞后一期
factor['EP']=factor.groupby('secID')['EP'].shift(1)
factor.dropna(inplace=True)
factor.head()
```

```python
#制作几个list
calendarlist=calendar[1:] #因为因子滞后一期，所以第一个日期可以从此剔除
calendarlist[:5] #观察calendarlist
secIDlist=np.unique(factor['secID'].values).tolist()
```

```python
#获取收益率
return_frame=pd.DataFrame()
for secID in secIDlist:
    return_frame=return_frame.append(DataAPI.MktEqumGet(secID=secID,ticker=u"",monthEndDate=u"",beginDate=calendarlist[0],endDate=calendarlist[-1],isOpen=u"1",field=u"secID,endDate,return",pandas="1"))
```

```python
#观察收益率DataFrame，各值是否合常理
return_frame.rename(columns={'endDate':'tradeDate'},inplace=True)#使column name和factor的一样便于操作
return_frame.sort_values('tradeDate').head()
```

```python
#合并因子和收益率DataFrame，用'left'合并 dropna==>排除某天停牌的
factor=pd.merge(factor,return_frame,how='left',on=['tradeDate','secID'])
factor.dropna(inplace=True)
factor.head()
```

```python
#保留此名为factor的DataFrame，以便后面模型调取数据，因此我们做一个copy()
factorframe1=factor.copy()
factorframe1.drop_duplicates(inplace=True)
```

```python
#对于某天，定义选股函数，直接return当天portfolio的收益率
def get_month_end_return1(factorframe1):
    df=factorframe1.copy()
    #选取前20EP的股票
    df=df.sort_values('EP',ascending=False)
    signal=df.head(20) 
    ret=(signal['return']*signal['EP']/signal['EP'].sum()).sum() #按因子加权投资，得到portfolio收益率
    return ret
```

```python
#得到此策略的monthly累积收益率DateFrame(定义2010-06-30开始交易，起始收益率为1),之后再plot。
finalreturn1=factorframe1.groupby('tradeDate').apply(get_month_end_return1)
finalreturn1=pd.Series({calendar[0]:0}).append(finalreturn1)
finalreturn1=finalreturn1+1
finalreturn1=finalreturn1.to_frame('simple_EP')
finalreturn1=finalreturn1.cumprod()
finalreturn1.head()
```

2.按照行业对因子进行选股，达到分散风险对冲的作用
即，对于每个交易每个行业都按照simple_EP来选股，选择行业前10
因子，我们需要获得每只股票所对应的行业以及行业权重

```python
last_month_date=calendar[-1].replace('-','')  #获取整个回测时间的最后一个交易日
industry=DataAPI.EquIndustryGet(industryVersionCD='010303',intoDate=last_month_date,field='secID,industryName1',pandas='1')  #按照申万一级行业获取每只股所对应的行业
industry.head()
```

为了计算行业权重，获取在时间上沪深300指数个股的权重，接着进行计算。

```python
#获取所有时间沪深300个股在整个市场中的权重
stock_weight=DataAPI.IdxCloseWeightGet(ticker='000300',beginDate=calendar[0],endDate=calendar[-1],field='consID,effDate,weight')
stock_weight.rename(columns={'effDate':'tradeDate','consID':'secID'},inplace=True) #保持列名一致
stock_weight.head()
```

```python
#合并股票行业和个股权重
stock_weight=pd.merge(stock_weight,industry,how='left',on='secID')
stock_weight.head()
```

```python
#定义对于一个特定时间，行业权重的计算函数
def cal_industry_weight(benchmark_weight):
    ben_wei_tod=benchmark_weight.copy()
    benchmark_indus_weight=ben_wei_tod.groupby('industryName1')['weight'].sum()             
    benchmark_indus_weight=benchmark_indus_weight.reset_index()
    benchmark_indus_weight.rename(columns={'weight':'industry_weight'},inplace=True)
    benchmark_indus_weight['industry_weight']=benchmark_indus_weight['industry_weight']/benchmark_indus_weight['industry_weight'].sum()
    return benchmark_indus_weight
```

```python
#获取时间上每个行业的权重
industry_weight=stock_weight.groupby('tradeDate').apply(cal_industry_weight)
industry_weight.reset_index(drop=False,inplace=True)
del industry_weight['level_1']
industry_weight.head()
```

```python
#获取所需DataFrame,时间，股票代码，所在行业，行业权重
factorframe2=factor.copy()
factorframe2=pd.merge(factorframe2,industry,how='left',on='secID')
factorframe2=pd.merge(factorframe2,industry_weight,on=['tradeDate','industryName1'])
factorframe2.drop_duplicates(inplace=True)
factorframe2.head()
```

```python
#对于某天某行业，定义交易函数，选取表现前10的股票，按照因子值的大小进行加权投资(直接返回该行业当天收益率)。
def one_day_one_indus(myframe):
    day_indus_frame=myframe.copy()
    day_indus_frame=day_indus_frame.sort_values('EP',ascending=False)
    day_indus_frame=day_indus_frame.reset_index(drop=True)
    signal=day_indus_frame.head(10)
    signal['weight']=(signal['EP']/signal['EP'].sum())*signal['industry_weight']
    signal['actual_return']=signal['return']*signal['weight']
    return sum(signal['actual_return'].values)
```

```python
#定义每月末总体收益率函数
def get_month_end_return2(finalframeoftoday):
    monthreturnframe=finalframeoftoday.copy()
    returnseries=monthreturnframe.groupby('industryName1').apply(one_day_one_indus)
    return returnseries.sum()
```

```python
#获取此策略下的收益率
finalreturn2=factorframe2.groupby('tradeDate').apply(get_month_end_return2)
finalreturn2=pd.Series({calendar[0]:0}).append(finalreturn2)
finalreturn2=finalreturn2+1
finalreturn2=finalreturn2.to_frame('EP_by_industry')
finalreturn2=finalreturn2.cumprod()
finalreturn2.head()
```

3.EP并不能完全代表公司股价上涨的概率
因为通常，市盈率和公司的流通市值是有相关性的(市盈率=市值/净利润率)(非因果关系)。
那么也就是说如果把市值看成自变量，则对EP的大小存在影响，则我们将EP进行改进
假设此关系呈线性关系，我们将EP对于Market Value进行回归，将残差作为新的因子，则可以消除市值的影响(中性化市值)
然后根据中性化后的因子，选取表现好的前20只股票

```python
#获取市值data，同样的，为了将市值和我们的因子值对应，对于每只股票将市值后移一个月的单位
market_value=pd.DataFrame()
for td in calendar:
    temp_market_value=DataAPI.MktEqudGet(secID=secIDlist,ticker=u"",tradeDate=td,beginDate='',endDate='',isOpen="",field=u"secID,tradeDate,marketValue",pandas="1")
    market_value=market_value.append(temp_market_value)
market_value['marketValue']=market_value.groupby('secID')['marketValue'].shift(1)
market_value.dropna(inplace=True)
market_value.head()
```

```python
#合并
factorframe3=pd.merge(factor,market_value,how='left',on=['secID','tradeDate'])
factorframe3.drop_duplicates(inplace=True)
factorframe3.head()
```

```python
#对于某天，定义中性化市值的函数
def neutralize_ep(factor_of_one_day):
    myfactor=factor_of_one_day.copy()
    model=sm.OLS(myfactor['EP'].values,myfactor['marketValue'].values).fit()  #线性回归
    myfactor['neutralized_factor']=myfactor['EP'].values-model.fittedvalues #residual
    return myfactor
```

```python
#获取处理过后的因子DataFrame
factorframe3=factorframe3.groupby('tradeDate').apply(neutralize_ep)
factorframe3=factorframe3.reset_index(drop=True)
factorframe3.head()
```

```python
#定义某天直接根据中性化之后的因子进行选股的函数
def get_month_end_return3(factorframe3):
    df=factorframe3.copy()
    #选取前20EP的股票
    df=df.sort_values('neutralized_factor',ascending=False)
    signal=df.head(20) 
    ret=(signal['return']*signal['neutralized_factor']/signal['neutralized_factor'].sum()).sum()
    #按因子加权投资，得到portfolio收益率
    return ret
```

```python
finalreturn3=factorframe3.groupby('tradeDate').apply(get_month_end_return3)
finalreturn3=pd.Series({calendar[0]:0}).append(finalreturn3)
finalreturn3.fillna(method='ffill',inplace=True)
finalreturn3=finalreturn3+1
finalreturn3=finalreturn3.to_frame('EP_MarketValue_neutralization')
finalreturn3=finalreturn3.cumprod()
finalreturn3.head()  #获取portfolio最终收益率
```

4.前3中策略均为选取前xxx的股票，无法做到完全分散unsystematic risk
因此，第四种策略基于第三种中性化之后的EP因子，对于全股进行投资，极大分散风险，做线性规划的简单优化

def optimize_foward_return(factorframe4oftoday):
    alpha=factorframe4oftoday['neutralized_factor']

结论:可以看出市盈率倒数EP因子在这几年表现较好，之前无明显变化，符合我们的预期。
需要注意得是回测也只是刻舟求剑，对过去适应可能只代表对未来短时间内同样适应。
因为用单因子对市场进行拟合分析会存在欠拟合的问题，环境变化太快，只用一个因子是不可能模拟经济环境的变化的。
单因子预测对于不远的过去回测，对于不远的将来才存才一定的参考价值，预测能力较强。时间过长则没有任何意义。
通过近几年的表现，发现EP因子表现较好，都比上证综指表现好太多，其中表现最好的是对于市值中性化之后的EP因子。
