# 天风长期择股策略4, 内含剔除次新, 剔除行业, ticker2secID

1. roe>0.1且roe<0.6, 经营性现金流为正,PE<50,考虑使用DataAPI.MktStockFactorsOneDayProGet
2. roe连续2季度回升, 考虑使用DataAPI.FdmtIndiRtnPitGet
3. 净利润增速单季度回升,毛利率单季度回升,考虑使用DataAPI.FdmtISGet,毛利率=(营业总收入-营业总成本)/营业总收入
4. 负债率<0.6,商誉比小于0.1,DataAPI.FdmtBSGet,负债率=负债总额/资产总额,商誉比=商誉/净资产(所有者权益)
5. PE历史分位<0.5, 滚动10年的PE分位, 考虑用for循环和DataAPI.MktStockFactorsDateRangeProGet同时使用
6. 上市时间超过1年, 考虑使用remove_new
7. 行业剔除金融业, 考虑使用DataAPI.EquIndustryGet, 金融业对应行业代码是01030110
8. 调仓时剔除停牌标的
9. 2009-05-04到2018-07-20年化24.1%, 每年5\9\11月第一天调仓, 因为4/30,8/31,10/31一季报/中报/三季报均会披露完

```python
#最大回撤期
def Cal_maxDraw(date,portf_value):
    backmax = portf_value.cummax()
    drawdown = portf_value/backmax-1
    index2 = drawdown.argmin()
    index1 = list(backmax).index(backmax[index2])
    return([date[index1],date[index2]])
Cal_maxDraw(bt.tradeDate,bt.portfolio_value)
```
