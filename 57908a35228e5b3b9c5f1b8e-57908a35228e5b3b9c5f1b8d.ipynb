`一.分位数去极值`
====

在分位数去极值中，比如可以把前后5%的分位数点，将大于或小于分位点的值删除掉，这种方法简单粗暴的一刀切，可能会牺牲部分正常的数据。当数据量较大的时候，随便扔点也不在乎，用吧。

```python
# 分位数去极值
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

se = np.random.randint(30, size=15)
plt.plot(se, 'r')
print "Series的值：" +  str(se)
print "5分位的值:  " + str(np.percentile(se, 90))

se[se>np.percentile(se, 90)] = np.percentile(se, 90)  #将极值赋值为分位点的值, 红色的是含有异常值的序列
print se
plt.plot(se, 'g')
```


`二.3倍标准差法`
====

当数据的分布是类正态分布的时候，可以使用3倍标准差的方法,如下图所示,将正负3$\sigma$之外的数据定义为异常值：

![](https://upload.wikimedia.org/wikipedia/commons/thumb/8/8c/Standard_deviation_diagram.svg/350px-Standard_deviation_diagram.svg.png)

下面的代码中有好几种3倍标准差法的实现，大同小异，其中scipy包的winsorize方法应该是最为方便的.

```python
# 3倍标准差法
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns

S=pd.Series(np.random.normal(size=200))
S[50] = 15   # 加入异常值
plt.plot(S, 'r')  #红色的曲线显示原始数据

S = S[~((S-S.mean()).abs()>3*S.std())]

#使用zscore函数
df = pd.DataFrame(np.random.randn(100, 3))
from scipy import stats
df[(np.abs(stats.zscore(df)) < 3).all(axis=1)]

plt.plot(S, 'b')

#使用scipy包的winsorize方法
from scipy.stats import mstats
result = mstats.winsorize(S.values, limits=0.05)

```


`三.中位数去极值`
====
使用中位数去极值的时候，发现相关的资料比较少，有些参考资料认为这种方法在很多方面要优于3倍标准差法，主要收集到下面三种方式，欢迎补充：

1. 中值滤波，使用中值滤波非常有可能改变原有位置的值，在[wiki](https://zh.wikipedia.org/wiki/%E4%B8%AD%E5%80%BC%E6%BB%A4%E6%B3%A2%E5%99%A8)中举的例子可以明显看到
2. 中位数去极值，直接使用与中位数的相对距离去除异常值
3. MAD方法，与上面提到的3倍标准差法很相似，但是用的是中位数，而不是均值。

```python
#中值滤波
#使用中值滤波会改变原有的数据，这种滤波特性在图像领域可以接受（一个像素点多几个灰度人眼不一定能分辨出来），但是其他领域可能需要打个问号了

import matplotlib.pyplot as plt
import scipy.signal as sps

raw = np.zeros(100)

# 加入一些异常点
raw[54:57] = 1
raw[23:24] = 1
raw[91] = -1
raw[76:80] = -1 # too big to be outlier

plt.axis([0, len(raw), -1.5, 1.5])

# figure(1)
plt.plot(raw, 'bo', label='raw data')
plt.show()

# filter outliers
filtered = sps.medfilt(raw, kernel_size=7)  # 更改窗口大小，可以看到过滤的效果
plt.plot(filtered, 'r', label='Filtered profile')

plt.axis([0, len(raw), -1.5, 1.5])
plt.legend(loc=3, ncol=2, borderaxespad=0.)
plt.show()
```

```python
#中位数去极值一种方法，这个方法没有找到比较规范的文档，只看到代码，但是代码比较简单

from scipy.signal import medfilt
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns

S=pd.Series(np.random.normal(size=200))
S[50] = 15   # 加入异常值
S[100] = -10
plt.plot(S, 'r')


#代码来源：http://stackoverflow.com/questions/11686720/is-there-a-numpy-builtin-to-reject-outliers-from-a-list
def reject_outliers(data, m =3):
    """Here I have replace the mean with the more robust median and the standard deviation with the absolute distance to the median. 
      I then scaled the distances by their (again) median value so that m is on a reasonable relative scale."""
    d = np.abs(data - np.median(data))
    mdev = np.median(d)
    s = d/mdev if mdev else 0.
    
    return data[s<m]

result = reject_outliers(S)

plt.plot(result, 'g')

```

```python
#代码来源： http://stackoverflow.com/questions/22354094/pythonic-way-of-detecting-outliers-in-one-dimensional-observation-data
#这个方法应该是参考某一篇论文，这个例子里面，作者将MAD的方法与分位数去极值做了详细的对比，可以看到分位数算法本身的局限，和中位数的优势。
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def main():
    for num in [10, 50, 100, 1000]:
        # Generate some data
        x = np.random.normal(0, 0.5, num-3)

        # Add three outliers...
        x = np.r_[x, -3, -10, 12]
        plot(x)

    plt.show()

def mad_based_outlier(points, thresh=3.5):
    if len(points.shape) == 1:
        points = points[:,None]
    median = np.median(points, axis=0)
    diff = np.sum((points - median)**2, axis=-1)
    diff = np.sqrt(diff)
    med_abs_deviation = np.median(diff)

    modified_z_score = 0.6745 * diff / med_abs_deviation

    return modified_z_score > thresh

def percentile_based_outlier(data, threshold=95):
    diff = (100 - threshold) / 2.0
    minval, maxval = np.percentile(data, [diff, 100 - diff])
    return (data < minval) | (data > maxval)

def plot(x):
    fig, axes = plt.subplots(nrows=2)
    for ax, func in zip(axes, [percentile_based_outlier, mad_based_outlier]):
        sns.distplot(x, ax=ax, rug=True, hist=False)
        outliers = x[func(x)]
        ax.plot(outliers, np.zeros_like(outliers), 'ro', clip_on=False)

    kwargs = dict(y=0.95, x=0.05, ha='left', va='top')
    axes[0].set_title('Percentile-based Outliers', **kwargs)
    axes[1].set_title('MAD-based Outliers', **kwargs)
    fig.suptitle('Comparing Outlier Tests with n={}'.format(len(x)), size=14)

main()
```
