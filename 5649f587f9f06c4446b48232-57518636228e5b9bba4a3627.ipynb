# Q&A【ROIC】

* 刚看了你在社区的求助帖，有下面一些问题

> 1. 此date_list只是一个简单的日历并没有在程序中发挥作用；

> 2. EBITToTOR['EBITToTOR']`[s]` 第二个方括号表示索引，源程序并没有设置索引

> 3. python严格区分大小写 IC['TShEquity'][`S`] 是不对的

> 4. 首先是没有import Series; 其次 ROIC = Series(ROIC) Series参数必须是可序列化的 , 你有些API读出的并不是一条记录。也就是说你ROIC[s] = (EBITToTOR['EBITToTOR'][s] * tRevenue['tRevenue'][s] * (1-0.25)) / (IC['TShEquity'][s] + IC['LTBorr'][s] + IC['minorityInt'][s]) 算出来的并不是一个值 ， 而是一个序列。 你需要对你之前取出的数据进行去重处理。请参看drop_duplicates 函数

> 5. accout.cash 今日是开盘前账户的现金 ， 但并不会因为你卖出股票后cash 发生变化。推荐用account.referencePortfolioValue

> 6. 关于ROIC 是很常用的估值因子，不过公式里面的 
* > 'TShEquity' 已经包括了少数股东权益 ；
* >　有息负债　并不简单等于长期负债
