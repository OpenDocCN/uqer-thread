StockScreener(<筛选条件表达式>) + [自定义股票列表]：股票筛选器，根据各个因子的数据，筛选股票列表，其中的因子用Factor.<factor_name>来实现，用起来很方便。支持5种方法进行筛选：

1）value_range(lbound, ubound): 筛选因子值处于上下界之间的证券（包含两端）。
2）pct_range(lbound, ubound): 筛选因子值处于百分比上下分位点之间的证券（包含两端），默认升序排列。
3）num_range(lbound, ubound): 筛选因子值处于上下界序号之间的证券（包含两端），默认升序排列。
4）nlarge(n): 筛选因子值最大的n只证券。
5）nsmall(n): 筛选因子值最小的n只证券。

1）ROIC增速（投入资本金回报率）>WACC增速（加权平均资本成本）

由于因子限制，用EBITToTOR代替ROIC，用DebtsAssetRatio代替WACC，如下：

EBITToTOR:息税前利润与营业总收入之比。计算方法: 息税前利润与营业总收入之比=(利润总额+利息支出-利息收入)/ 营业总收入，如果没有利息支出，用财务费用代替，以上科目使用的都是TTM的数值。属于盈利能力和收益质量类因子。
DebtsAssetRatio:债务总资产比（Debt to total assets）。计算方法：债务总资产比=负债合计/总资产。属于偿债能力和资本结构类因子。

2）存货周转率<20%

使用InventoryTRate,如下：
InventoryTRate：存货周转率（Inventory turnover rate）。计算方法：存货周转率=营业成本（TTM）/存货。属于营运能力类因子。

3）经营活动现金流>0

使用OperCashGrowRate，如下：
OperCashGrowRate：经营活动产生的现金流量净额增长率。计算方法：经营活动产生的现金流量净额增长率=(今年经营活动产生的现金流量净额（TTM）/去年经营活动产生的现金流量净额（TTM）)-1。属于现金流指标因子。

5）净利润净资产收益率上升

用NetProfitGrowRate，NetAssetGrowRate，如下：
NetAssetGrowRate：净资产增长率（Net assets growth rate）。计算方法：净资产增长率=(今年股东权益/去年股东权益)-1。属于成长能力类因子。
NetProfitGrowRate：净利润增长率（Net profit growth rate）。计算方法：净利润增长率=(今年净利润（TTM）/去年净利润（TTM）)-1。属于成长能力类因子。
