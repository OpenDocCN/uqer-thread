```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import datetime as dt
import CAL.PyCAL as C_cal # CAL provides classes and functions dealing with date, time, and calendars. 

sns.set_style('white')
pd.set_option('display.width', 200)
cal = C_cal.Calendar('China.SSE')
```

```python
begin_Date = '20180801' # 开始日期
end_Date = (dt.datetime.today() - dt.timedelta(days=1)).strftime('%Y%m%d') # 截止日期

IndustryEval=DataAPI.MktIndustryEvalGet(beginDate=end_Date,endDate=end_Date,industryID=u"",PEType="2",field=u"",pandas="1")

# industryMarketValue	int	行业总市值
IndustryEval['industryMarketValue']=IndustryEval['industryMarketValue']/sum(IndustryEval['industryMarketValue'])
average_PB=sum(IndustryEval['PB']*IndustryEval['industryMarketValue'])
average_PE=sum(IndustryEval['PEValue']*IndustryEval['industryMarketValue'])
IndustryEval['average_PB']=average_PB
IndustryEval['average_PE']=average_PE

del IndustryEval['industryIncome']
del IndustryEval['industryAttrP']
                
# IndustryEval.set_index('industryName', inplace=True)
IndustryEval.sort_values(by=['PEValue','PB'], axis=0, ascending=[True,True], inplace=True)
IndustryEval.reset_index(drop=True,inplace=True)

print((end_Date[0:4]+'年'+end_Date[4:6]+'月'+end_Date[6:8]+'日')+'的 行业板块 PB PE 的指标 ；\n'+
     'average_PB:'+str(average_PB)+'    average_PE:'+str(average_PE))

IndustryEval[['PB','average_PB']].plot.bar(figsize=(16,5),grid = True,stacked=False)
IndustryEval[['PEValue','average_PE']].plot.bar(figsize=(16,5),grid = True,stacked=False)

pd.set_option('display.max_columns', None) # 显示所有列
pd.set_option('display.max_rows', None) # 显示所有行
IndustryEval
```
