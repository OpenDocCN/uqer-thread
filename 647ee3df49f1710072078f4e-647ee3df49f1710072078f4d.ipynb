```python
import numpy as np
import pandas as pd
from CAL.PyCAL import *
cal=Calendar('China.SSE')
import seaborn as sns
import statsmodels.api as sm
import matplotlib.pyplot as plt
from datetime import timedelta
from statsmodels import regression
```

```python
def candle(Index,n):
    df= Index.copy()
    df['candle_up']= df['highestIndex']-df['high']
    df['candle_up_mean']= df['candle_up'].rolling(n).mean()
    df['std_candel_up']= df['candle_up']/df['candle_up_mean']
    
    '''
    df['candle_up']= 0.0
    for i in range(len(df)):
        df['candle_up'].iloc[i]= df['highestIndex'].iloc[i]-df['high'].iloc[i]
    df['std_candel_up']=0.0
    for i in range(n,len(df)):
        df['std_candel_up'].iloc[i]= df['candle_up'].iloc[i]/np.mean(df['candle_up'].iloc[i-n:i])
    '''
    df['std_candel_up_std']=df['std_candel_up'].rolling(20).std()
    return df
```

```python
###############
#计算RSRS指标
###############
def RSRS(Index,Date,beta_list,N,M,S1,S2):
    buy_flag = 0
    sell_flag = 0
    
    
    #############################
    #得到最高指数序列与最低指数序列
    #############################
    beta_out =  pd.DataFrame(index=range(M-1),columns=['alpha','beta','tradeDate'])
    start = Index[Index['tradeDate']==Date].index.values
    start = int(start[0])
    last_start =start-N
    Index_use = Index[last_start+1:start+1]
    #计算未来10天的收益率
    #IndexCompare = Index[start:start+10]
    #rt10= IndexCompare['closeIndex'].iloc[-1]/IndexCompare['closeIndex'].iloc[0]-1
    ########################################
    #对最高指数序列与最低指数序列进行线性回归
    ########################################
    high = np.array(Index_use['highestIndex'])
    low = np.array(Index_use['lowestIndex'])
    #beta = np.polyfit(low, high,1)
    
    #high= np.array(Index_use['high'])
    #low = np.array(Index_use['low'])
    
    X=sm.add_constant(low)
    ols_out=sm.OLS(high,X).fit()
    
    alpha= ols_out.params[0]
    beta= ols_out.params[1]
    rsquared= ols_out.rsquared_adj
    
    ##############################
    #将该天的RSRS指标插入beta_list中
    ##############################
    
    #beta_insert=pd.DataFrame(index=[M-1],columns=['alpha','beta','tradeDate'])
    #beta_insert['alpha'][M-1]=beta[1]
    #beta_insert['beta'][M-1]=beta[0] 
    
    beta_insert=pd.DataFrame(index=[M-1],
                             columns=['alpha','beta','rsquared','tradeDate'])
    beta_insert['alpha'][M-1]=alpha
    beta_insert['beta'][M-1]=beta
    beta_insert['rsquared'][M-1]=rsquared
    
    beta_insert['tradeDate'][M-1]=Index['tradeDate'][start]
    temp = beta_list.drop(0)
    beta_out['alpha'] = np.array(temp['alpha'])
    beta_out['beta'] = np.array(temp['beta'])
    beta_out['tradeDate'] = np.array(temp['tradeDate'])
    beta_out = pd.concat([beta_out,beta_insert],axis=0)
    ######################################################
    #利用beta_list中的数据进行标准化得到改天的beta数据的标准化
    ######################################################
    
    beta_standard = np.sqrt(1.0*(M-1)/M)*(beta - np.mean(beta_out['beta'][len(beta_out)-M:len(beta_out)]))/np.std(beta_out['beta'][len(beta_out)-M:len(beta_out)])*rsquared*beta
    
    
    MA20 = np.mean(Index['closeIndex'][start-21:start+1])
    MA20_2 = np.mean(Index['closeIndex'][start-26:start-4])
    
    MA5= np.mean(Index['closeIndex'][start-6:start+1])
    MA5_2= np.mean(Index['closeIndex'][start-8:start-1])
    
    #计算过去n天的波动率
    pct1 = np.std(Index['closeIndex'].iloc[start-20:start])
    
    #hurst指数
    #hurst= Index['hurst'].iloc[start]
    
    #蜡烛线上std
    candle= Index['std_candel_up_std'].iloc[start]
    
    ##############
    #买卖信号判断
    ##############
    if(beta_standard>S1 and MA20_2<MA20):
        #((MA20<MA5 and MA20_2>MA5_2)or(MA20<MA5 and MA20_2<MA5_2))):
        buy_flag = 1
    elif(beta_standard<S2 or candle>1.7):
        sell_flag = 1
    '''
    print(Index['tradeDate'].iloc[start], Index['tradeDate'].iloc[start+1])
    print(beta_out[-1:])
    print('adj_beta:',beta_standard)
    print('return:',pct1)
    print('sell_flag:',sell_flag)
    print('buy_flag:',buy_flag)
    '''
    
    return buy_flag,sell_flag,beta_out,beta_standard,beta,pct1,candle

###########################################
#计算RSRS指标在开始交易的M天的指标序列
###########################################
def RSRS_Init(Index,Date,N,M):
    #############################
    #得到最高指数序列与最低指数序列
    #############################
    beta_list = pd.DataFrame(index=range(0),columns=['alpha','beta','rsquared','rt','tradeDate'])
    start = Index[Index['tradeDate']==Date].index.values    
    start = int(start[0])
    start = start-M+1
    
    
    ########################################
    #计算RSRS指标M天的指标序列
    ########################################
    for i in range(M):
        ########################################
        #对最高指数序列与最低指数序列进行线性回归
        ########################################
        last_start =start-N
        Index_use = Index[last_start:start]
        
        high = np.array(Index_use['highestIndex'])
        low = np.array(Index_use['lowestIndex'])
        
        #high = np.array(Index_use['high'])
        #low = np.array(Index_use['low'])
        
        X=sm.add_constant(low)
        ols_out=sm.OLS(high,X).fit()
        alpha= ols_out.params[0]
        beta= ols_out.params[1]
        rsquared= ols_out.rsquared_adj

        #beta = np.polyfit(low, high,1)
        
        ###################################
        #将新得到的线性回归系数插入beta_list
        ###################################
        L = len(beta_list)
        beta_insert=pd.DataFrame(index=[L],
                                 columns=['alpha','beta','rsquared','tradeDate'])
        #beta_insert['alpha'][L]=beta[1]
        #beta_insert['beta'][L]=beta[0]
        
        beta_insert['alpha'][L]=alpha
        beta_insert['beta'][L]=beta
        beta_insert['rsquared'][L]=rsquared
        
        beta_insert['tradeDate'][L]=Index['tradeDate'][start]
        beta_list = pd.concat([beta_list,beta_insert],axis=0)
        start = start + 1
    return beta_list
```
