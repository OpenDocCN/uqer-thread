#### 1.趋势判断
##### 第一步：任选一只股票，画出它的在2014年的收盘价变化曲线。在这里小散取的是000002，大名万科A，简称WKA
##### 第二步：对收盘价数据进行OLS回归，将得到的Y=aX+b直线绘在图中。

```python
# Import libraries to find linear trend and plot data
from statsmodels import regression
import statsmodels.api as sm
import scipy.stats as stats
import scipy.spatial.distance as distance
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from CAL.PyCAL import * 

start = '20140101'
end = '20141230'
asset=DataAPI.MktEqudGet(ticker='000002',beginDate=start,endDate=end,field=['tradeDate','closePrice'],pandas="1").set_index('tradeDate')['closePrice'] # 使用Dataapi获取万科A的收盘价

# 绘制收盘价变化曲线
dates = asset.index

asset.plot(figsize=(12,8))
# 对收盘价数据进行直线拟合
X = np.arange(len(asset))
x = sm.add_constant(X)
model = regression.linear_model.OLS(asset, x).fit() # 使用OLS计算y=ax+b对应的a和b
a = model.params[0]
b = model.params[1]
Y_fit = X * b + a
plt.plot(X, Y_fit, 'r', alpha=0.9);
plt.ylabel('Price')
plt.legend(['WKA', 'Trendline']);
```

观察得到的图形，万科A在2014年的价格走势可以看作线性回归线（即趋势线）与随机噪声的叠加。

#### 2.自相关检验

#### 小散在决定买入股票时，通常会参考之前交易日的价格，定量表示为：
$$P_t = X + P_{1, t-1} + \epsilon$$
#### 式中，$P_{1, t-1}$表示上一交易日价格 ，$X$代表其他影响股价变化的因素，$\epsilon$表示随机噪声。如果上式成立，则表明股价具有自相关性。自相关的解释请自行百度，谷歌，必应。
#### 有效市场假说认为，投资者以线性方式而不是累计方式对信息作出反应，今天的价格只能由今天的新闻引起，今天的收益与昨天无关，收益是互相独立的，不具有自相关性。显然小散并不认可这个假说。


下面我们来直观的观察一下自相关特性,构建两个函数：

函数1 ：

```python
def generate_autocorrelated_values(N):
    X = np.zeros(N)
    for i in range(N-1):
        X[i+1] = X[i] + np.random.normal(0, 1) #看这里
    return X

for i in range(10):
    X = generate_autocorrelated_values(100)
    plt.plot(X)
plt.xlabel('$t$')
plt.ylabel('$X_t$');
```

函数2：

```python
def generate_autocorrelated_values(N):
    X = np.zeros(N)
    for i in range(1, N-1):
        past_returns = X[i] - X[i-1]                   #看这里
        X[i+1] = X[i] + past_returns + np.random.normal(0, 1)   #看这里
    return X

for i in range(10):
    X = generate_autocorrelated_values(10)
    plt.plot(X)
plt.xlabel('$t$')
plt.ylabel('$X_t$');
```

PS：不能有效区分两个函数类型的自行面壁10分钟。

#### 根据上面的解释，具有动量特征的股票，其股价通常是自相关的。那么自相关如何检测呢？
#### 在这里，小散采用了一种叫做单位根检验的办法，在万能的Python里调用statsmodels模块下的adfuller函数，可以很容易的实现这个操作。赶紧测一下之前提到的万科A是不是自相关的。

```python
from statsmodels.tsa.stattools import adfuller
print 'WKA'
X1=asset
_, pvalue, _, _, _, _ = adfuller(X1)
if pvalue > 0.05:
    print '自相关'
else:
    print '非自相关'
```

##### 看来运气不错，万科A是自相关的，可以愉快的应用动量规律进行判断了。

#### 3.趋势反转

##### 与动量效应相对的是反转效应，指过去一段时间收益率较高的股票在未来获得的收益率将会低于过去收益率较低的股票。
##### 在这里，时间范围的选取尤为重要。回顾之前提到的万科A的走势，2015年整体呈上升趋势并伴有阶段性的趋势反转。
##### 长线投资者可以选择在年初买入，并持续持有。小散是一个没有耐心的人，恨不得每天操作一次，怎样才能科学的进行低抛高吸的工作呢？
##### 答案是：捕捉反转信号。这里提供一种方法，供大家参考。

```python
(asset - Y_fit).plot(figsize=(12,8))
plt.hlines(np.mean(asset - Y_fit), 0, 300, colors='r')
plt.hlines(np.std(asset - Y_fit), 0, 300, colors='r', linestyles='dashed')
plt.hlines(-np.std(asset - Y_fit), 0, 300, colors='r', linestyles='dashed')
plt.xlabel('Time')
plt.ylabel('Dollar Difference');
```
