# 指数成份股调整带来的投资机会

## 指数成份股调整
----
每年的6月和12月，中证指数公司会对重要A股指数的成分股进行调整，包括沪深300指数、上证180等。而随着指数成分股的调入调出，跟踪该指数的基金包括ETF基金、LOF基金等也会对基金的持仓做出相应的调整，从而有可能会对即将调入调出的股票价格产生重要影响。指数的成分股定期调整，一般经历3个阶段：考察期结束日、大约1个半月后的“调整名单公布日”、2到3周后的“调整实施日”。“调整名单公布日”那天，交易所会在官网公布调整的成分股名单。只要名单一公布，各大基金经理就必须在2到3周之内调仓完毕。“调整实施日”当天，指数就以新的成分股进行运作。

```python
def get_quotation(start_date,end_date,universe):
    df_eq = pd.DataFrame()
    for i in range(30,len(universe),30):
        df_tmp = DataAPI.MktEqudAdjGet(secID=universe[i-30:i],beginDate=start_date,endDate=end_date,
                                    field=u"secID,tradeDate,openPrice",isOpen=1,pandas="1")
        df_eq = pd.concat([df_eq,df_tmp],axis=0)
    df_tmp = DataAPI.MktEqudAdjGet(secID=universe[i:],beginDate=start_date,endDate=end_date,
                                field=u"secID,tradeDate,openPrice",isOpen=1,pandas="1")
    df_eq = pd.concat([df_eq,df_tmp],axis=0)
    return df_eq
```

3.时间窗口分析
--

```python
def get_window_yield(unstack_yield,events,window):
    #获取事件窗口收益
    events = events.reset_index()
    def get_time_list(window,events):
    # 得到每个事件的时间窗口
        minus_period = '-' + str(window) + 'B'
        period = str(window) + 'B'
        time_list = [[cal.advanceDate(events['publistDate'][i],minus_period,BizDayConvention.Unadjusted),cal.advanceDate(events['publistDate'][i],period,BizDayConvention.Unadjusted)] for i in range(len(events['publistDate']))]
        return time_list
    
    time_list = get_time_list(window,events)
    offset = range(-1*window,window+1,1)
    event_equity_list = []
    for i in range(len(events)):
        # print time_list[i],len(unstack_yield.ix[time_list[i][0].strftime('%Y-%m-%d'):time_list[i][1].strftime('%Y-%m-%d'),events.iloc[i]['consID']])
        df_tmp = pd.DataFrame(unstack_yield.ix[time_list[i][0].strftime('%Y-%m-%d'):time_list[i][1].strftime('%Y-%m-%d'),events.iloc[i]['consID']].values,columns=[events.iloc[i]['secID_publistDate']],index = offset)
        event_equity_list.append(df_tmp)
    event_equity_window  = pd.concat(event_equity_list,axis=1)
    return event_equity_window
```

```python
def get_mean_outperform_yield(event_equity_window_yield,event_benchmark_window_yield):
    #获取窗口内平均超额收益
    cols = list(event_equity_window_yield[(event_equity_window_yield > 0.4) | (event_equity_window_yield < -0.3)].dropna(how='all',axis=1).columns) #删除异常值，一天内的open2open收益最高也就是30%多点，存在借壳上市的公司会有很高的收益，删除这样的样本点
    for col in cols:
        del event_equity_window_yield[col]
        del event_benchmark_window_yield[col]
    event_equity_window_cumyield = (1 + event_equity_window_yield).cumprod(axis=0) / (1 + event_equity_window_yield).cumprod(axis=0).ix[0] # 累计收益，对齐T+0点
    event_benchmark_window_cumyield = (1 + event_benchmark_window_yield).cumprod(axis=0) / (1 + event_benchmark_window_yield).cumprod(axis=0).ix[0]
    event_outperform_window_yield = (event_equity_window_cumyield - event_benchmark_window_cumyield).mean(axis=1) #超额收益累计平均值
    return event_outperform_window_yield
```

```python
def event_window_plot(open2open_yield,benchmarkYield,events,window,legend):
    ## 画出各事件时间窗口收益
    event_equity_window_yield = get_window_yield(open2open_yield,events,window)
    event_benchmark_window_yield = get_window_yield(benchmarkYield,events,window)
    event_outperform_window_yield = get_mean_outperform_yield(event_equity_window_yield,event_benchmark_window_yield)
    fig = plt.figure(figsize=(10,6))
    ax = fig.add_subplot(111)
    ax.plot(event_outperform_window_yield.index,event_outperform_window_yield.values,'dodgerblue',linewidth=2.5)
    plt.xlim(-window,window)
    plt.ylim(event_outperform_window_yield.min(),event_outperform_window_yield.max())
    ax.plot([0,0],[event_outperform_window_yield.min(),event_outperform_window_yield.max()],'crimson',linewidth=1.5,linestyle ='--')
    ax.plot([1,1],[event_outperform_window_yield.min(),event_outperform_window_yield.max()],'crimson',linewidth=1.5,linestyle ='--')
    ax.set_title(legend, fontproperties=font, fontsize=18)
```

4.多空策略
---
显然对于调入和调出指数的股票，我们很自然的会想用多空策略，看多调入的股票，然后看空调出的股票。这样可以规避掉很多其他的因素对股票收益的影响，而且能进HS300的股票大多都可以融资融券的，所以这个是具有现实意义的。
<br>
这里我们构建多空策略，持仓周期为名单公布日到生效日，即名单公布后的隔天开盘买入，生效日当天开盘卖出，多空都为等权。

```python
##获取时间窗口每日收益
event_into_yield = get_window_yield(df_eq_ret,df_into_valid,21)
event_out_yield = get_window_yield(df_eq_ret,df_out_valid,21)
##获取每个指数成份股调整的日期与对应股票列表序列
into_list = df_into_valid[['publistDate','secID_publistDate']].groupby('publistDate').apply(lambda x:x['secID_publistDate'].tolist())
out_list = df_out_valid[['publistDate','secID_publistDate']].groupby('publistDate').apply(lambda x:x['secID_publistDate'].tolist())
```

```python
##计算每一次指数成分股策略在持仓日期的累计收益
trade_date_num = [cal.bizDatesNumber(item[1],item[0]) for item in date_tuple]
outcome_summary = dict()
for i,idx in enumerate(into_list.index):
    outcome_summary[idx] = (1 + event_into_yield[into_list[idx]].ix[2:trade_date_num[i]-1].mean(axis=1)).cumprod() - (1 + event_out_yield[out_list[idx]].ix[2:trade_date_num[i]-1].mean(axis=1)).cumprod()
outcome_summary = pd.DataFrame(outcome_summary)
```

```python
outcome_summary.head(5)
```

```python
cum_out_yield = [outcome_summary[col].dropna().iloc[-1] for col in outcome_summary.columns] ##计算每次操作的最终收益
```

```python
print len([item for item in cum_out_yield if item > 0]),len(cum_out_yield),1.0*len([item for item in cum_out_yield if item > 0]) / len(cum_out_yield)
```

如上，我们可以知道10年以来一共14次指数成份股调整，其中12次使用我们的多空策略可以获得正收益，胜率高达85.7%。我们做出累计净值走势，如下所示。

```python
def netValue_plot():
    position_date = [cal.bizDatesList(item[1],item[0])[2:] for item in date_tuple] # 获取持仓日期
    cum_ret_summary = []
    ##计算每一个持仓周期内的累计收益
    for i,col in enumerate(outcome_summary.columns):
        cum_ret_summary.append(pd.Series((1 +outcome_summary[col]).dropna().tolist(),index=position_date[i]))
    cum_ret_summary = pd.concat(cum_ret_summary,axis=0) 
    ##将每个持仓周期的累计收益再时间轴上累计起来
    for i in range(1,len(position_date)):
        cum_ret_summary[position_date[i]] = cum_ret_summary[position_date[i-1]][-1] * cum_ret_summary[position_date[i]]
    cum_ret_series = cum_ret_summary.ix[cal.bizDatesList('2009-12-16','2016-06-30')]
    cum_ret_series = cum_ret_series.fillna(method = 'ffill')
    ##画净值曲线图
    fig = plt.figure(figsize=(12,10))
    ax  = fig.add_subplot(111)
    ax.plot([ item.toDateTime() for item in list(cum_ret_series.index)],cum_ret_series.values,'dodgerblue',linewidth=2.5)
    ax.legend([u"净值走势"],prop = font,loc='upper left',frameon= False)
    ax.set_title(u"指数成份股调整多空策略(不考虑交易成本)",fontproperties=font, fontsize=18)
```

5.总结与进一步展望
----
### `总结`
本文通过分析指数成份股变动事件，首先以成份股调整名单公布时间为基准，进行事件时间窗口分析，结果表明，调入股票在调整名单公布后平均有一个显著的涨幅，但10个交易日左右，又开始下跌，而且整个事件窗口收益不太稳定；而调出股票在调整名单公布后平均有一个显著的跌幅，并且在20个交易日内稳定持续的下跌。接着，依据这些信息，我们编写了基于调入调出股票的多空策略，持仓周期为调整名单发布日至生效日，等权设置多空。结果，总体操作的胜率在85.7%，10年至16年间取了70%多的累计收益。
### `进一步研究`
上面的结果还比较喜人，可以看出在大盘比较稳定的时候，这样多空操作基本都是能赚钱。而这些赚钱的原因，我们也大致知道是因为指数基金需要买入调入的股票卖出调出股票。既然如此，更加严谨的做法应该是测算大致会有多少这样的资金量买入需要调入的股票，因为一只股票可能同时在几个指数，主流指数都有指数基金，再加这个股票在各指数所占权重，以及该指数整体基金规模，就可以大致计算出有多少这样的资金会买入。然后再考虑，这些资金是否足够大，能给股票带来显著影响。调出股票同样考虑。
