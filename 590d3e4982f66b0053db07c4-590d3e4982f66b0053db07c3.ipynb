# 获取分红数据
首先获取A股历史现金分红数据，再计算其登记日的股息率。

```python
import pandas as pd
import numpy as np
from dateutil.parser import parse
import datetime
import matplotlib.pylab as plt
from CAL.PyCAL import *
import statsmodels.api as sm 
from sklearn.linear_model import LogisticRegression
cal = Calendar('China.SSE')
```

```python
def get_universe(trade_date):
    """
    获取动态股票列表
    """
    eq_msg = DataAPI.EquGet(equTypeCD=u"A",secID=u"",ticker=u"",listStatusCD=u"L,S,DE,UN",
               field=u"secID,listDate,delistDate",pandas="1")
    eq_msg = eq_msg.fillna(datetime.datetime.today().strftime('%Y-%m-%d'))
    eq_msg = eq_msg[(eq_msg['listDate']<trade_date) & (eq_msg['delistDate']>trade_date)]
    return eq_msg['secID'].tolist()
```

```python
universe = DataAPI.EquGet(equTypeCD=u"A",secID=u"",ticker=u"",listStatusCD=u"L,S,DE,UN",field=u"secID",pandas="1")['secID'].tolist()#包括退市的股票
```

```python
div_summary = DataAPI.EquDivGet(exDivDate="",eventProcessCD=u"",secID=universe,ticker=u"",
                  beginDate=u"20070420",endDate=u"20160520",field=['secID','perCashDiv','recordDate','endDate','publishDate'],pandas="1")
div_summary = div_summary.dropna(subset=['recordDate','publishDate'])
div_summary.set_index(np.arange(0,len(div_summary)),inplace=True)
div_summary = div_summary.fillna(0)
return_rate=pd.DataFrame(columns=['secID','ratio','endDate','recordDate','publishDate'])
return_rate[['secID','endDate','publishDate','recordDate']]=div_summary[['secID','endDate','publishDate','recordDate']]

for i in np.arange(0,len(div_summary)):
    
    publish_price = DataAPI.MktEqudGet(secID = div_summary.secID[i] , tradeDate= div_summary.recordDate[i] , field=u"closePrice",pandas="1")
    return_rate.ratio[i] =  div_summary.perCashDiv[i] / publish_price.closePrice[0]*100.0
    
return_rate=return_rate[return_rate['ratio']>0]
```

```python
high_ratio=return_rate.groupby(pd.qcut(return_rate['ratio'],10)).size()

high_ratio
```

## 由历史数据可知，股息率大于2.14%的概率为10%，由此可以定义股息率大于2.14%的个股为高股息个股，即发生高分红。

```python
return_rate=return_rate[return_rate['ratio']>2.14]
```

# 建立模型
目标： 通过11月份已知的信息来预测当年年终该股票是否会发生高分红。
考虑变量（信息）
1. rePS(每股未分配利润)
2. EPS(原文用的TTM，可是需要Pro权限，在此用low版替代)
3. 总市值
4. 20个交易日平均价
5. 是否新股（上市250个交易日以内）
6. 上市时间
7. 年中是否有高分红
8. 去年是否有高分红

```python
def get_point_date(start,end):
    point_date = []
    for year in range(start,end):
        point_date.append(str(year)+'-11-01')
    return point_date
def get_point_universe(point_date):
    available_universe_dict = {}
    for date in point_date:
        available_universe_dict[date] = get_universe(date)
    return available_universe_dict
point_date = get_point_date(2007,2015) ## 得到2007年11月1至2014年11月1日
universe_dict = get_point_universe(point_date)
```

```python
def get_factor(trade_date,universe):
    trade_date = cal.adjustDate(trade_date).strftime('%Y%m%d')
    begin_date = cal.advanceDate(trade_date,'-20B').strftime('%Y%m%d')
    begin_date_rePS = cal.advanceDate(trade_date,'-2M').strftime('%Y%m%d')
    ##得到EPS
    fundament = DataAPI.MktStockFactorsOneDayGet(tradeDate=trade_date,secID=universe,ticker=u"",
                                    field=u"secID,tradeDate,EPS",pandas="1")
    '''fundament['Total_PS'] = fundament['CapitalSurplusFundPS'] + fundament['RetainedEarningsPS']''' # 每股资本公积金与每股留存收益之和
    ##得到总市值
    quotation = DataAPI.MktEqudGet(tradeDate=trade_date,secID=universe,ticker=u"",
                                   beginDate=u"",endDate=u"",isOpen="",field=u"secID,marketValue",pandas="1")
    quotation['marketValue'] = quotation['marketValue']  / 1e+08 ##总市值(亿)
    
    ##计算平均股价
    quotation_window = DataAPI.MktEqudGet(tradeDate=u"",secID=universe,ticker=u"",
                                   beginDate=begin_date,endDate=trade_date,isOpen="",field=u"secID,closePrice",pandas="1")
    # print quotation_window,begin_date,trade_date
    quotation_window = quotation_window.groupby(['secID']).mean().reset_index().rename(columns = {'closePrice':'meanPrice'})
    
    ##计算rePS
    rePS=DataAPI.FdmtIndiPSPitGet(secID=universe,endDate=trade_date,beginDate=begin_date_rePS,field=u"secID,rePS",pandas="1")
    
    ##计算上市时间
    eq_list = DataAPI.EquGet(equTypeCD=u"A",secID=universe,ticker=u"",listStatusCD=u"",field=u"secID,listDate",pandas="1")
    eq_list['listYear'] = eq_list['listDate'].map(lambda x: cal.bizDatesNumber(x,trade_date)/250.0) 
    
    ##设置是否为次新
    eq_list['ifNew'] = eq_list['listYear'].map(lambda x:1 if x<=1 else 0)
    
    summary_df = pd.merge(fundament[['secID','tradeDate','EPS']],quotation[['secID','marketValue']],on=['secID'],how='inner')
    summary_df = pd.merge(summary_df,eq_list[['secID','listYear','ifNew']],on=['secID'],how='inner')
    summary_df = pd.merge(summary_df,quotation_window[['secID','meanPrice']],on=['secID'],how='inner')
    summary_df = pd.merge(summary_df,rePS[['secID','rePS']],on=['secID'],how='inner')
    ##年中是否有过高分红
    mid_transfer_list = return_rate[(return_rate['secID'].isin(universe)) & (return_rate['endDate']== (trade_date[0:4]+'-06-30'))]['secID'].tolist()
    summary_df['MidHasDiv'] = summary_df['secID'].map(lambda x: 1 if x in mid_transfer_list else 0)
    ##去年是否有过高分红
    last_transfer_list = return_rate[(return_rate['secID'].isin(universe)) & (return_rate['endDate']== str(int(trade_date[0:4])-1)+'-12-31')]['secID'].tolist()
    summary_df['LastHasDiv'] = summary_df['secID'].map(lambda x: 1 if x in last_transfer_list else 0)
    
    return summary_df
```

```python
def get_total_summary(universe_dict):
    """
    获取所有样本点汇总
    """
    total_summary_df = pd.DataFrame()
    for k,v in universe_dict.items():
        summary_df = get_factor(k,v)
        result = return_rate[(return_rate['publishDate']>k) & (return_rate['publishDate']<str(int(k[0:4])+1)+'-05-01')]
        summary_df['isDiv'] = summary_df['secID'].map(lambda x:1 if x in result['secID'].tolist() else 0)
        total_summary_df = pd.concat([total_summary_df,summary_df],axis=0)
    return total_summary_df
```

```python
total_summary_df = get_total_summary(universe_dict)
total_summary_df = total_summary_df.dropna()
total_summary_df = total_summary_df.reset_index(drop=True)
```

## 对所有因子进行标准化处理

```python
#标准化处理
def zscore(group):
    return (group-group.mean())/group.std()
total_summary_df.ix[:,2:-1]=total_summary_df.ix[:,2:-1].apply(zscore)
```

得到解释变量与被解释变量构成的矩阵，其中`rePS`是每股未分配利润、`EPS`是每股收益、`marketValue`是总市值单位为亿、`listYear`是上市时间单位年、`ifNew`是虚拟变量表示是否是新股、`meanPric`是20个交易日平均价、`MidHasDiv`是虚拟变量表示年中是否有过高分红、`LastHasDiv`是虚拟变量表示去年是否有过高分红，一共7个解释变量；`isDiv`是被解释变量表示年终是否真的高分红。

```python
total_summary_df.head()
```

```python
total_summary_df['intercept'] = 1.0##加入常数项
total_summary_df.head()
```

```python
trans_col = list(total_summary_df.columns[2:-2]+total_summary_df.columns[-1:])#获取解释变量列

print trans_col,
```

```python
logit = sm.Logit(total_summary_df['isDiv'],total_summary_df[trans_col])
result = logit.fit()
#classifier = LogisticRegression()  # 使用类，参数全是默认的  
#classifier.fit(total_summary_df[trans_col], total_summary_df['isDiv'])  # 训练数据来学习，不需要返回值  
```

```python
logit_1 = sm.Logit(total_summary_df['isDiv'],total_summary_df[trans_col_1])
result_1 = logit_1.fit()
# classifier = LogisticRegression()  # 使用类，参数全是默认的  
# classifier.fit(total_summary_df[trans_col], total_summary_df['isTransfer'])  # 训练数据来学习，不需要返回值  
```

可以发现所有因子均显著，`EPS`、`rePS`、`LastHasDiv`、`marketValue`越大，高分红的概率越大，`meanPrice`、`listYear`越大，高分红概率越小。

# 样本内与样本外预测
---
依据得到的模型进行样本内外预测

```python
in_Sample_Predict = result_1.predict(total_summary_df[trans_col_1])
in_Sample_Result = total_summary_df[:]
in_Sample_Result['predict_probability'] = pd.Series([round(prob,3) for prob in in_Sample_Predict],index=in_Sample_Result.index)
in_Sample_Result.groupby(['tradeDate']).apply(lambda x:x.sort_values(['predict_probability'],ascending = False))[['secID','isDiv','predict_probability']]
```

如上便是所有样本内的情况,大家可以对比其预测高分红（predict_probability）与是否发生了高分红（isDiv）。以2015年年终的高分红实际情况，作为样本外点预测。

```python
out_Sample_Universe = get_universe('2015-11-01')
out_Sample_df = get_factor('2015-11-01',out_Sample_Universe)
out_Sample_df = out_Sample_df.dropna()
out_Sample_df['intercept'] = 1
out_Sample_df.head()
```
