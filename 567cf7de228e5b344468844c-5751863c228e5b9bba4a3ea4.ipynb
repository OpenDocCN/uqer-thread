```python
# 导入包
import pandas as pd
import numpy as np

import math
from datetime import datetime, timedelta
import matplotlib.pyplot as plt

def  hs300_weigth(ws):
    if     ws > 0.8:        return 1;
    elif    ws > 0.7:        return  0.8;
    elif    ws > 0.6:        return  0.7;
    elif    ws > 0.5:        return  0.6;
    elif    ws > 0.4:        return  0.5;
    elif    ws > 0.3:        return  0.4;
    elif    ws > 0.2:        return  0.3;
    elif    ws > 0.15:       return  0.2;
    
    return round(ws,2);


def  show_hs300_his_peb():
      
    start_date = '20070430'  # 开始日期
    #start_date = '20150430'  # 开始日期
    end_date = (datetime.today() - timedelta(days=1)).strftime('%Y%m%d')  # 截止日期
    data = DataAPI.TradeCalGet(exchangeCD=u"XSHG",beginDate=start_date,endDate=end_date,field=u"calendarDate,isWeekEnd",pandas="1")
    data = data[data['isWeekEnd'] == 1]
    Dates = map(lambda x: x[0:4]+x[5:7]+x[8:10], data['calendarDate'].values.tolist())

    PB_PE = pd.DataFrame(index=Dates, data=0, columns=['tradeDate','total PB','total PE'])
    for date in Dates:  
        universe = set_universe('HS300', date) 
        total = DataAPI.MktStockFactorsOneDayGet(tradeDate=date,secID=universe[:-1],field=['ticker','LCAP','LFLO','PB','PE'],pandas="1").set_index('ticker')
        length_total = total.shape[0]
        total.dropna(inplace=True)
    
        # 数据量不足80%时，不具有代表性
        if float(total.shape[0]) / length_total < 0.8:
            continue
    
        # 细节处理：负值极值处理
        total['PB'][total['PB'] < 0] = 0
        total['PE'][total['PE'] < 0] = 0
        pb = winsorize(total['PB'].to_dict())
        pe = winsorize(total['PE'].to_dict())
        total['PB'][pb.keys()] = pb.values()
        total['PE'][pe.keys()] = pe.values()
    
        # 权重 + 绘图
        total['wts1'] = map(lambda x,y:  hs300_weigth((10**x)/(10**y)), total['LFLO'].values.tolist(), total['LCAP'].values.tolist())
        total['wts2']= map(lambda x,y:x* (10**y), total['wts1'].values.tolist(),total['LCAP'].values.tolist())        
        total['wts'] = total['wts2'] / sum(total['wts2'])
        total['PE'] = total['PE'] * total['wts1']
        
        PB_PE.loc[date] = [date,sum(total['wts'] * total['PB']), sum(total['wts'] * total['PE'])] 

        
    PB_PE['average PB'] = PB_PE['total PB'].mean()
    PB_PE['average PE'] = PB_PE['total PE'].mean()

    #print PB_PE
    
    # 趋势线
    fig = plt.figure(figsize=(19,24))
    ax1 = fig.add_subplot(4,1,1)   
    ax1.grid(True)
    ax1.set_title(u'HS300 PE')
    ax1.set_ylabel('PE')
    ax1.plot(PB_PE[['total PE','average PE']])
    ax1.legend(['total PE','average PE'])     
    
    #生成日期
    xticks = ax1.get_xticks()
    xticks_label=[]
    for xpos in xticks:
       try:
            xticks_label.append(Dates[int(xpos)])
       except:
            continue    
    ax1.set_xticklabels(xticks_label)
    
    #xticks_label =map(lambda x:if x < len(ax1.get_xticks()):Dates[int(x)] ,ax1.get_xticks()
    
    ax3 = fig.add_subplot(4,1,3)
    ax3.grid(True)
    ax3.set_title(u'HS300 PB')
    ax3.set_ylabel('PB')
    ax3.plot(PB_PE[['total PB','average PB']],color='green')
    ax3.legend(['total PB','average PB'])
    ax3.set_xticklabels(xticks_label)
    
    # 直方图
    ax2 = fig.add_subplot(4,1,2)
    ax2.hist(PB_PE['total PE'],20,normed=1)
    ax2.set_xlabel('PE')
    ax2.set_ylabel('Probability')
    ax2.grid(True)
    ax2.set_title(u'HS300 PE Histogram')
    
    ax4 = fig.add_subplot(4,1,4)
    ax4.set_xlabel('PB')
    ax4.set_ylabel('Probability')
    ax4.grid(True)
    ax4.set_title(u'HS300 PB Histogram')
    ax4.hist(PB_PE['total PB'],20,normed=1,facecolor='green')
    return

show_hs300_his_peb();
```
