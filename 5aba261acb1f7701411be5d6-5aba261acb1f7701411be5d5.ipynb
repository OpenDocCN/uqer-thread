```python
import pandas as pd
import numpy as np
start = '2013-01-01'
end = '2018-03-20'
stk='600016.XSHG'
name='民生银行'
# for stk in ['600016.XSHG','600015.XSHG','601328.XSHG']:
# name=DataAPI.SecIDGet(ticker= stk.split('.'),field=u"secShortName",pandas="1").values[0][0]
universe = DynamicUniverse(stk)
benchmark = stk
# refresh_rate =Monthly(2)
refresh_rate =1
freq = 'd'    
max_history_window = 1

accounts = {
    'fantasy_account': AccountConfig(account_type='security', capital_base=10000000)
}

# 把回测参数封装到 SimulationParameters 中，供 quick_backtest 使用
sim_params = quartz.SimulationParameters(start, end, benchmark, universe, refresh_rate=refresh_rate,freq=freq, accounts=accounts, max_history_window=max_history_window)

# ---------------回测参数部分结束----------------
# 获取回测行情数据
data = quartz.get_backtest_data(sim_params)
# 运行结果
results = {}
# 调整参数快速回测
for pbegin in [0.1,0.15,0.2,0.25,0.3,0.35,0.4,0.45,0.5]:
    for pend in [0.1,0.15,0.2,0.25,0.3,0.35,0.4,0.45,0.5]:
        for bk_len in [1000]:
            param=[pbegin,pend,bk_len]
            # -----策略定义开始，这和常用的策略编写模式完全一样-----    
            def initialize(context):
                context.pbegin, context.pend, context.bk_len=param
                context.stk=stk
                # name=DataAPI.SecIDGet(ticker= context.stk.split('.'),field=u"secShortName",pandas="1").values[0][0]
                print name
                data=pd.read_csv('./银行详细因子/'+name+'_PB等.csv',index_col=0)
                #---------------------------------------------------
                #参数
                # context.pbegin=0.1
                # context.pend=0.1
                # context.bk_len=100
                ####################################################
                pbegin=context.pbegin
                pend=context.pend
                #---------------------------------------------------
                data.R=data.PB/(data.NPParentCompanyGrowRate+1.)
                bk_len=context.bk_len
                data['Rmax']=None
                data['Rmin']=None
                x0=np.array(data.R)
                x=np.array(data.R)
                xmin=np.array(data.R)
                xmax=np.array(data.R)
                for i in range(len(data.index)-bk_len-1):
                    x=x0[i:i+bk_len]
                    x=x.copy()
                    x.sort()
                    xmin[i+bk_len+1]=x[int(len(x)*pbegin)]
                    xmax[i+bk_len+1]=x[int(len(x)*(1.-pend))]
                data['Rmin']=xmin
                data['Rmax']=xmax
                context.data=data

            def handle_data(context):
                previous_date = context.previous_date.strftime('%Y-%m-%d')
                data=context.data
                p=1.-(data.R[previous_date]-data.Rmin[previous_date])/(data.Rmax[previous_date]-data.Rmin[previous_date])
                if p>1:
                    p=1.
                if p<0:
                    p=0.
                #-----------------------------------------------
                # if p>0.5:
                #     p=1.
                # else:
                #     p=0.
                #-----------------------------------------------
                account = context.get_account('fantasy_account')   
                account.order_pct_to(context.stk, p)
                # return    

            # 生成策略对象
            strategy = quartz.TradingStrategy(initialize, handle_data)
            # -------------------策略定义结束--------------------   

            # 开始回测
            bt, perf = quartz.quick_backtest(sim_params, strategy, data=data,refresh_rate=refresh_rate)
            # 保存运行结果
            results[str(param)] = {'max_drawdown': perf['max_drawdown'],
                              'sharpe': perf['sharpe'],
                              'alpha': perf['alpha'],
                              'annualized_return':perf['annualized_return'],
                            'benchmark_annualized_return':perf['benchmark_annualized_return']
                             }    
            print  param,perf['annualized_return']
data=pd.DataFrame(results).T
name=DataAPI.SecIDGet(ticker= stk.split('.'),field=u"secShortName",pandas="1").values[0][0]
data.to_csv('./PB个股策略结果/'+name+'1.csv')
```
