# 程序员学量化投资（七）：量价八阶律的简单实现

上篇文章我们讲到了利用成交量来选股，但是单独看成交量是很难判断股票走势，所以本次我们引入价格因子，通过量价关系来选股。

量价关系是一个重要的指标,量是价的先驱，而价格又可以影响成交，彼此互为消长就形成了周期性的循环。这个成交量与价格变化的周期我们称之为“八阶律”。

![img](http://n1.itc.cn/img8/wb/smccloud/fetch/2015/05/02/47025366053213947.JPEG)

本策略就是实现程序判断量价的增减状态和目前所处的量价阶段，从而根据行情的走势做出合理的买卖操作，实现盈利。

上面是以平安银行个股为例，回测12年至今的收益， 从图表可以看出这个策略相对稳健和保守，熊市以及平时可以保证基本收益，但是牛市表现较差，甚至跑不赢大盘。

所以我们再完善一下，把股票池设置成沪深300，这样我们把网撒大一点，规避风险，尽量弥补策略的不足。  最后牛市里的表现确实好了一点~
