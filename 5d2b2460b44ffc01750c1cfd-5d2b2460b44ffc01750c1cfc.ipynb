```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import datetime as dt
from CAL.PyCAL import * # CAL provides classes and functions dealing with date, time, and calendars. 

sns.set()
pd.set_option('display.width', 200)
cal = Calendar('China.SSE')
```

```python
data=DataAPI.MktEqudGet(ticker="600030",beginDate="20181001") #选取600030股票

a = data.closePrice
B = []
n = len(a)
for i in range(10, n):
    x5 = a[i-5:i].mean() #5日均线值
    x10 = a[i-10:i].mean() #10日均线值
    B.append(x5 > x10)

o = data.openPrice
m = len(B)
w = 0 #利润
cash = 1000000 #操作金额
amount = 0
PL = [] #利润w的数组

for i in range(1, m):
    k = i + 10
    if B[i-1] == 0 and B[i] == 1 and not amount:
        amount = cash // o[k] #买入份额
        cash -= o[k] * amount
    elif B[i-1] == 1 and B[i]==0 and amount:
        cash += o[k] * amount #卖出的金额
        amount = 0

    PL.append(cash + o[k] * amount)
   
plt.figure(figsize=(25, 7))
plt.plot(PL,color="green",label="Profit and Loss")
plt.xlabel("Date")
plt.ylabel("Price")
plt.show()

plt.figure(figsize=(25, 7))
plt.plot(a[10:], color="red",label="Profit and Loss")
plt.show()

B
```
