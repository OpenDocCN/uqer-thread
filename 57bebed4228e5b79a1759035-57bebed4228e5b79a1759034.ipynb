```python
import numpy as np
import pandas as pd
from pandas import DataFrame
# 获取沪深300成分股所有数据
Data = DataAPI.TradeCalGet(exchangeCD=u"XSHG",beginDate=u"20090101",endDate=u"20160620",field=u"calendarDate,isOpen",pandas="1")
isOpen = Data['isOpen']
calendarDate = Data['calendarDate'][isOpen == 1]
i = 0
for date in calendarDate:
    hs300_stock_data = DataAPI.IdxConsGet(secID=u"",ticker=u"000300",intoDate=date,isNew=u"",field=u"intoDate,consID",pandas="1")
    i = i + 1
    if i == 1:
        hs300_stock = list(hs300_stock_data['consID'])
        hs300_data = hs300_stock_data['consID']
    else:
        temp = list(hs300_stock_data['consID'])
        hs300_stock = list(set(hs300_stock).union(set(temp)))
        hs300_temp_data = hs300_stock_data['consID']
        hs300_data = pd.concat([hs300_data,hs300_temp_data],axis = 1)
hs300_data.columns = calendarDate
# 获取时间段内所有沪深300成分股的开盘价数据
Sec_Code = hs300_stock
# 获取因子数据
factors = list(['tradeDate,BLEV','tradeDate,CashRateOfSales','tradeDate,CashToCurrentLiability','tradeDate,DebtEquityRatio',
           'tradeDate,EPS','tradeDate,ETOP','tradeDate,FixAssetRatio','tradeDate,GrossIncomeRatio','tradeDate,NetProfitRatio','tradeDate,NPToTOR',
           'tradeDate,OperatingProfitRatio','tradeDate,ROA','tradeDate,ROE','tradeDate,AD','tradeDate,CoppockCurve','tradeDate,EMV6',
           'tradeDate,ADX','tradeDate,Aroon','tradeDate,DDI','tradeDate,MTM','tradeDate,MTMMA','tradeDate,PVT','tradeDate,UOS',
           'tradeDate,Hurst','tradeDate,EGRO','tradeDate,InvestCashGrowRate','tradeDate,NetAssetGrowRate','tradeDate,NetProfitGrowRate',
           'tradeDate,OperatingRevenueGrowRate','tradeDate,TotalAssetGrowRate','tradeDate,TotalProfitGrowRate'])    
for factor in factors:
    newdata = DataFrame()
    i = 0
    for SecID in Sec_Code:
        data = DataAPI.MktStockFactorsDateRangeGet(secID=SecID,beginDate=u"20090101",endDate=u"20160620",field=factor,pandas="1")
        i = i + 1
        if i == 1:
            newdata = pd.concat([newdata, data], axis = 1)
        else:
            newdata = newdata.merge(data, on = "tradeDate", how = "outer")
    newdata.columns = list(["Date"] + Sec_Code)
    filename = factor + '.csv'
    newdata.to_csv(filename)
```
