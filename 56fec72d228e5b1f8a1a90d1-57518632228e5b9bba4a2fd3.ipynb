## 一 问题背景

期权的波动率估计一般有两种方法：第一种是根据历史数据来预测未来的波动率。一般而言，期权对应的原生资产（例如股票）的波动率，都会存在显著的自相关性，因此需要使用时间序列分析（EWMA模型，GARCH模型等）进行建模和预测。

另外一种估计期权波动率的方法是计算隐含波动率。隐含波动率是通过市场上不同期限（记为T）和行权价格（记为K）的期权定价反推其对应的波动率，得到一个以T和K作为自变量的二维隐含波动率曲面。当要对一个具有特定期限或者波动率的新期权进行定价时，可以通过在隐含波动率曲面上进行插值来估算这个新期权相应的波动率。

这篇文章基于《C++ Design Patterns and Derivatives Pricing》第九章所讨论的隐含波动率求解的C++实现思路，来对隐含波动率的数值求解问题作简单讨论。

## 二 问题分析

从期权价格求解其隐含波动率通常是一个非线性方程求解问题，一般难以获得其闭式解。以不考虑分红的欧式看涨期权的Black-Scholes (BS) 风险中性公式定价为例：

$$ C(S,t) = S(t)*N(d_1)-K*exp{[-r(T-t)]*N(d_2)} $$

$$ d_1 = \dfrac{1}{\sigma\sqrt{(T-t)}}*[ln(\dfrac{S(t)}{K}+(r+\dfrac{1}{2}\sigma^2))] $$

$$ d_2 = d_1-\sigma\sqrt{T-t} $$

其中，$C(S,t)$ 为为t时刻欧式看涨期权价格，S为t时刻的股价，$N(*)$为标准正态分布的累积分布函数，$d_1$和$d_2$均为累积分布函数的积分上限，r为无风险利率，T为期权的期限，$exp{-r(T-t)}$为贴现因子，K为行权价格，$\sigma$为波动率。在BS公式中，r和$\sigma$均假设为常数。

定价公式中共有五个参数：t时刻的股价，执行价格，利率，期限和波动率。如果考虑分红，则共有六个参数。

根据上述定义，问题可简化表述为，给定除波动率外的参数及期权价格，求波动率。

$$ C = f(\sigma, S(t), K, r, d, T) $$

## 三 二分法和牛顿法介绍

二分法的思路相对简单。在求解方程f(x)=0时，如果函数f(x)是连续函数，当f(a)$<$0 且 f(b)$>$0，或f(a)$>$0 且 f(b)$<$0，则区间 $[a,b]$ 中必定存在f(x)的解。在迭代求解过程中，我们不断计算区间中点的函数值来决定新区间，当达到我们需要的精度，则迭代停止。

设f(a)$<$0 且 f(b)$>$0，则：

(1) 计算 $f(\dfrac{a+b}{2})$；

(2) 如果 $f(\dfrac{a+b}{2})=0$，则迭代停止；

(3) 如果 $f(\dfrac{a+b}{2})$小于0，则新区间为$[\dfrac{a+b}{2},b]$；

(4) 如果 $f(\dfrac{a+b}{2})$大于0，则新区间为$[a, \dfrac{a+b}{2}]$；

(5) 当 $f(\dfrac{a+b}{2})$的绝对值小于我们需要的精度，迭代停止。

二分法的优点在于它必定会收敛到方程的解，且不要求函数是可导的（梯度下降法，共轭梯度法或牛顿法等均需要计算函数的导数）；缺点其收敛速度慢，一般只适用于低维的非线性方程组求解。

牛顿法的迭代是基于函数f(x)在Xn处的一阶泰勒展开式：

$$ f(X_{n+1}) = f(X_{n}) + f'(X_{n})(X_{n+1}-X_{n}) $$

令$f(X_{n+1})=0$ , 得到

$$ X_{n+1} = X_{n} - \dfrac{f(X_{n})}{f'(X_{n}} $$

按照上式进行迭代，直到$f(X_{n+1})$的绝对值小于我们需要的精度，则迭代结束。

牛顿法的优点在于收敛速度较快（二次收敛），因此适用于高维的方程组，且在解附近收敛性较好；缺点在于如果起始点离解较远，存在不收敛的可能。因此通常梯度下降法会和牛顿法联合使用，先用梯度下降法接近解的区域，再使用牛顿法。

另外，牛顿法也广泛应用于优化问题中。求连续可导函数的极值，等价于求解方程:

$$ f'(X_{n+1})=0 $$

对函数的一阶导数f’(x)在Xn处进行一阶泰勒展开，且令$f(X_{n+1})=0$ , 得到：

$$ X_{n+1} = X_{n} - \dfrac{f'(X_{n})}{f''(X_{n})} $$

对于高维函数，牛顿法的每一次迭代都需要重新计算二阶导数矩阵${f''(X_{n})}$（汉森矩阵）。由于其运算量较大，在实际问题中，一般采用拟牛顿法（BFGS算法等），来对汉森矩阵进行近似计算，加快运算速度。

## 四 二分法模块的C++ 设计模式

在使用二分法求解隐含波动率的时候，因为不同的期权，或者不同的定价理论会给出不同的定价公式，我们需要求解的非线性方程组也会有所不同。因此为了保证二分法模块的代码重用性，函数f(x)将从模块外部传入。

书中介绍了两种不同的设计模式。第一种称为模板设计模式（template pattern），其思路是在模块中定义二分法的计算流程，而定价公式作为纯虚函数传入模块中（此时二分法模块也成为一个抽象类），而具体的定价公式在派生类中实现。书中第七章曾使用这个设计模式对奇异期权进行定价。

模板设计模式的缺点主要有两个：1. 因为二分法需要频繁调用f(x)来计算区间中点值来进行比较，因此把f(x)设定为纯虚函数本身会降低程序效率。每次调用虚函数，都需要在虚函数表(virtual function table)中进行搜索；2. 另一个把f(x)设定为虚函数的问题在于，在个别情况下，f(x)需要继承别的类来获得一些函数所需要的特性，这样f(x)所在的派生类需要进行多重继承（multiple inheritance），会增加代码复杂性。

第二种实现模式称为模板化（templatization ）。模板化允许编译器在编译时对函数的代码进行内联和优化，从而避免了频繁调用函数f(x)带来的效率问题。而模板化的缺点在于会增加代码数量。

在二分法模块实现中，模板化能够有效避免频繁调用函数的效率问题，同时由于二分法计算的代码较短，代码量增加的代价是可以接受的。因此书中采用了模板化的设计模式来实现二分法计算。

为了允许函数f(x)保存自己的参数，这里把它声明为一个函数对象（function object）。在C++中，函数对象是重载了函数调用运算符的类。此时类既可以像函数一样被调用，也可以像类一样保存自己的数据成员。例如，在利用BS公式的模块计算期权价格的时候，我们希望这个模块能够保存一些给定的参数（利率r，分红率d，期限T，股价S和行权价格K），同时当我们调用这个模块时，能够根据这些给定的参数，和新的波动率值计算期权的价格。

根据上述讨论，我们来看调用BS定价公式计算期权的BSCallClass的头文件和源文件：















```python
//
//
//                  BSCallClass.h
//
//

#ifndef BS_CALL_CLASS_H
#define BS_CALL_CLASS_H
class BSCall
{

public:

    BSCall(double r_, double d_,
                     double T, double Spot_,
                     double Strike_);

    double operator()(double Vol) const;


private:

    double r;
    double d;
    double T;
    double Spot;
    double Strike;

};
#endif
```

```python
//
//
//                      BSCallClass.cpp
//
//

#include <BSCallClass.h>
#include <BlackScholesFormulas.h>

BSCall::BSCall(double r_, double d_,
               double T_, double Spot_,
               double Strike_)
               :
               r(r_),d(d_),
               T(T_),Spot(Spot_),
               Strike(Strike_)
{}

double BSCall::operator()(double Vol) const
{
    return BlackScholesCall(Spot,Strike,r,d,Vol,T);
}
```

上面的代码相对简单。头文件中，public部分声明了类的构造函数BSCall，和重载了函数调用运算符。这里用const来对函数调用运算符进行了修饰，保证了这个类只会根据输入的波动率（vol变量）进行计算，而不会修改数据，提高了代码的安全性。而计算期权价格所需要的其它参数均在private部分进行了声明。

源文件中对函数对象进行了定义。当输入新的波动率数值，就会调用BlackScholesCall函数计算相应的期权值。

二分法的模板类代码头文件如下：

```python
//
//
//              Bisection.h
//
//


template<class T>
double Bisection(double Target, 
                 double Low, 
                 double High,
                 double Tolerance,
                 T TheFunction) 
{

    double x=0.5*(Low+High);
    double y=TheFunction(x);
    
    do
    {
       
        if (y < Target)
            Low = x;

        if (y > Target)
            High = x;

        x = 0.5*(Low+High);

        y = TheFunction(x);
    
    }
    while 
        ( (fabs(y-Target) > Tolerance) );

    return x;
}
```

Bisection.h代码开头是template<class T>，表示将声明的类是模板类。当我们把BSCall作为参数，在main程序中以Bisection<BSCall>调用模板类时，所有的T在编译阶段都会被替换为BSCall。

另外需要注意的是，模板类的代码都必须放在头文件，而不能放在源文件。因为源文件无法预编译，所以模板类也无法在编译阶段实现内联，而造成程序错误。

最后看二分法调用的主程序SolveMain1.cpp

## 四 牛顿法的C++ 设计模式

牛顿法和二分法的实现差别主要在于牛顿法需要输入两个函数（f(x)及其一阶导数）。一个解决办法是分别输入两个函数对象。但这样要对同样一组的参数进行两次初始化，还要保证它们一致；而且就逻辑而言，函数值和函数的一阶导数值描述的是同一个事物的两个性质，把它们分为两个对象并没有反映它们的实际关系。

另一个解决办法是直接声明一个derivative函数。但实际上，BSFormula中已经有了我们需要的期权对波动率的一阶导数（vega），此时定义一个名为derivative，重新计算函数一阶导数并不是很好的实现方式。

以上的问题可以通过类成员函数指针来解决，先来看NewtonRaphson.h的代码：

```python
//
//
//                  NewtonRaphson.h
//
//



template<class T, double (T::*Value)(double) const, double (T::*Derivative)(double) const >
double NewtonRaphson(double Target,
                 double Start,
                 double Tolerance,
                 const T& TheObject)
{
    
    double y = (TheObject.*Value)(Start);
   
    double x=Start;

    while ( fabs(y - Target) > Tolerance )
    {
        double d = (TheObject.*Derivative)(x);
         
        x+= (Target-y)/d;

        y = (TheObject.*Value)(x);
   
    }

    return x;
}
```

在NewtonRaphson.h中，我们声明了两个指向类成员函数的指针（T::Value和T::Derivative）。类成员函数指针的的作用和函数指针相似，区别在于它只能指向特定类的成员函数。语法上，类成员函数的指针需要加上“类名::”；而在调用类成员函数指针时，需要使用“对象名.类成员函数指针名”的语法。

另外，为了避免运行时指针定义带来的时间消耗，这里把函数指针设定为模板参数而不是函数参数。这意味着编译器会把它按模板参数处理，在编译阶段就确定它们的类型进行内联。

牛顿法调用BS公式代码计算新的期权价格和二分法类似，区别只在于除了函数值以外，牛顿法还需要知道函数的一阶导数值（vega)。下面是牛顿法调用BS公式计算期权值的BSCallTwo类的头文件和源文件代码：

## 五 模板化设计模式的优缺点讨论

上面介绍了数值求解隐含波动率的数值方法（二分法和牛顿法），和在C++中利用模板化的设计模式进行实现。

和模板设计模式比较，模板化的设计模式使得对象类型在编译期间决定的，而模板设计模式中的纯虚函数则是代码运行区间决定的。所以模板化的优点在于它不需要在执行时判断执行哪段代码，而且因为对象类型确定，编译器可以对代码进行适当的优化。而模板化的缺点在于编译时间变慢，且因为在每次使用的模板的地方都会对代码进行编译，因此程序中有了同一份代码的多份拷贝。因此执行文件会变大。如果使用多个模板的话，代码量将大幅增加。

模板化的另一个可能存在的问题在于，编译器不会编译那些模板化代码中那些没有被调用的函数。因此如果一个方法有错误，而目前又未被调用，可能要很久以后才发现存在问题。解决这个问题的办法是先对代码进行调试，再进行模板化。
