# 

期权的波动率交易中，我们需要先对未来的已实现波动率进行估计，然后和市场上期权所隐含的波动率进行相应的交易。预测波动率之前，需要对过去的波动率水平进行度量。本文中，研究度量历史波动率的各种方法。

---

# 1. 波动率的定义和度量

最简单的已实现波动率的计算，就是标的资产收益率的标准差。进一步，为了真实体现标的资产的价格波动，人们定义了几个不同的波动率计算方法（以下的波动率函数均来自于论文： Drift-Independent Volatility Estimation Based on High, Low, Open, and Close Prices. http://papers.ssrn.com/sol3/papers.cfm?abstract_id=229190）.

--- 

## 1.1 Close to Close波动率：

$$ \sigma = \sqrt{\frac{1}{N-1}\sum_{i=1}^N \bigg[\bigg(\ln\frac{C_i}{O_i} + \ln\frac{O_i}{C_{i-1}}\bigg)-\frac{1}{N}\sum_{i=1}^N\bigg(\ln\frac{C_i}{O_i} + \ln\frac{O_i}{C_{i-1}}\bigg)\bigg]^2}  \\ =\sqrt{\frac{1}{N-1}\sum_{i=1}^N \bigg[\bigg(\ln\frac{C_i}{C_{i-1}}\bigg)-\frac{1}{N}\sum_{i=1}^N\bigg(\ln\frac{C_i}{C_{i-1}}\bigg)\bigg]^2}$$

$C,O,L,H$分别表示一个时间间隔内的收盘价、开盘价、最低价和最高价(下同)。

- 优点：抽样特性容易理解；偏差容易校正；可以简单地转换成带有日回报项的公式形式
- 缺点：没有充分利用已有数据的信息，收敛速度很慢

## 1.2 Parkinson波动率：

$$ \sigma = \sqrt{\frac{1}{4N\ln{2}}\sum_{i=1}^N \bigg(\ln\frac{H_i}{L_i}\bigg)^2}$$

- 优点：使用日内极差构造波动率很有意义，相比于Close to Close，提供了更全面的信息
- 缺点：只适合用于估计几何布朗运动过程中的波动率；尤其不能处理价格趋势和价格跳空；系统性地低估波动率

## 1.3 Rogers Satchell Yoon 波动率：

$$ \sigma = \sqrt{\frac{1}{N}\sum_{i=1}^N \bigg(\ln\frac{H_i}{C_i}\bigg)\bigg(\ln\frac{H_i}{O_i}\bigg) +\bigg(\ln\frac{L_i}{C_i}\bigg)\bigg(\ln\frac{L_i}{O_i}\bigg) }$$

- 优点：允许价格趋势效应的存在
- 缺点：无法处理价格跳空

## 1.4 Garman Klass 波动率：

$$ \sigma = \sqrt{(\sigma_o^\prime)^2 - 0.383(\sigma_c^\prime)^2 + 1.364 \sigma_P^2 + 0.019 \sigma^2_{RS}} $$
其中
$$(\sigma_o^\prime)^2 = \frac{1}{N}\sum_{i=1}^N \bigg(\ln\frac{O_i}{C_{i-1}}\bigg)^2 \\
(\sigma_c^\prime)^2 = \frac{1}{N}\sum_{i=1}^N \bigg(\ln\frac{C_i}{O_{i}}\bigg)^2 \\
\sigma_P^2 = {\frac{1}{4N\ln{2}}\sum_{i=1}^N \bigg(\ln\frac{H_i}{L_i}\bigg)^2} \\
\sigma_{RS}^2 = {\frac{1}{N}\sum_{i=1}^N \bigg(\ln\frac{H_i}{C_i}\bigg)\bigg(\ln\frac{H_i}{O_i}\bigg) +\bigg(\ln\frac{L_i}{C_i}\bigg)\bigg(\ln\frac{L_i}{O_i}\bigg) }
$$

- 优点：效率比close to close 高出很多；充分利用常用的价格信息
- 缺点：比Parkinson估计量的偏差更大

## 1.5  Yang Zhang 波动率：

$$ \sigma = \sqrt{\sigma_O^2+k\sigma_C^2+(1-k)\sigma_{RS}^2} $$
其中
$$\sigma_O^2 = \frac{1}{N-1}\sum_{i=1}^N\bigg[\ln\frac{O_i}{C_{i-1}}-\frac{1}{N}\bigg(\sum_{i=1}^N\ln\frac{O_i}{C_{i-1}}\bigg)\bigg]^2 \\
\sigma_C^2 = \frac{1}{N-1}\sum_{i=1}^N\bigg[\ln\frac{C_i}{O_{i}}-\frac{1}{N}\bigg(\sum_{i=1}^N\ln\frac{C_i}{O_{i}}\bigg)\bigg]^2 \\
k = \frac{0.34}{1.34+\frac{N+1}{N-1}}
$$

- 优点：具有最小估计误差的波动率估计；能够同时处理价格漂移和价格跳空；在可用数据的使用上最为有效
- 缺点：当价格过程由跳空主导时，其性能会降低到和close to close估计量的差不多

# 2. 波动率的计算

```python
from CAL.PyCAL import *
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('white')
```

```python
## 计算一段时间标的的 Close to Close 历史波动率，返回值包括以下不同周期的波动率：
# 一周，半月，一个月，两个月，三个月，四个月，五个月，半年，九个月，一年，两年
def getHistVolatilityC2C(secID, beginDate, endDate):
    cal = Calendar('China.SSE')
    spotBeginDate = cal.advanceDate(beginDate,'-520B',BizDayConvention.Preceding)
    # spotBeginDate = Date(2006, 1, 1)
    begin = spotBeginDate.toISO().replace('-', '')
    end = endDate.toISO().replace('-', '')
    
    fields = ['tradeDate', 'preClosePrice', 'closePrice', 'settlePrice', 'preSettlePrice']
    security = DataAPI.MktFunddGet(secID=secID, beginDate=begin, endDate=end, field=fields)
    security['dailyReturn'] = security['closePrice']/security['preClosePrice']    # 日回报率
    security['u'] = np.log(security['dailyReturn'])      # u2为复利形式的日回报率
    security['tradeDate'] = pd.to_datetime(security['tradeDate'])
    
    periods = {'hv1W': 5, 'hv2W': 10, 'hv1M': 21, 'hv2M': 41, 'hv3M': 62, 'hv4M': 83, 
               'hv5M': 104, 'hv6M': 124, 'hv9M': 186, 'hv1Y': 249, 'hv2Y': 497}
    # 利用方差模型计算波动率
    for prd in periods.keys():
        tmp = pd.rolling_std(security['u'], window=periods[prd])*math.sqrt(252.0)
        security[prd] = np.round(tmp, 5)
    
    security = security[security.tradeDate >= beginDate.toISO()]
    security = security.set_index('tradeDate')
    return security

## 计算一段时间标的的Parkinson历史波动率，返回值包括以下不同周期的波动率：
# 一周，半月，一个月，两个月，三个月，四个月，五个月，半年，九个月，一年，两年
def getHistVolatilityParkinson(secID, beginDate, endDate):
    cal = Calendar('China.SSE')
    spotBeginDate = cal.advanceDate(beginDate,'-520B',BizDayConvention.Preceding)
    # spotBeginDate = Date(2006, 1, 1)
    begin = spotBeginDate.toISO().replace('-', '')
    end = endDate.toISO().replace('-', '')
    
    fields = ['tradeDate', 'highestPrice', 'lowestPrice']
    security = DataAPI.MktFunddGet(secID=secID, beginDate=begin, endDate=end, field=fields)
    security['dailyAmplitude'] = security['highestPrice']/security['lowestPrice']    # 日回报率
    security['u'] = np.power(np.log(security['dailyAmplitude']), 2)/4.0/np.log(2.0)      # u2为复利形式的日回报率
    security['tradeDate'] = pd.to_datetime(security['tradeDate'])
    
    periods = {'hv1W': 5, 'hv2W': 10, 'hv1M': 21, 'hv2M': 41, 'hv3M': 62, 'hv4M': 83, 
               'hv5M': 104, 'hv6M': 124, 'hv9M': 186, 'hv1Y': 249, 'hv2Y': 497}
    # 利用Parkinson模型计算波动率
    for prd in periods.keys():
        tmp = np.sqrt(pd.rolling_mean(security['u'], window=periods[prd]))*math.sqrt(252.0)
        security[prd] = np.round(tmp, 5)
    
    security = security[security.tradeDate >= beginDate.toISO()]
    security = security.set_index('tradeDate')
    return security

## 计算一段时间标的的Rogers Satchell Yoon历史波动率，返回值包括以下不同周期的波动率：
# 一周，半月，一个月，两个月，三个月，四个月，五个月，半年，九个月，一年，两年
def getHistVolatilityRogersSatchellYoon(secID, beginDate, endDate):
    cal = Calendar('China.SSE')
    spotBeginDate = cal.advanceDate(beginDate,'-520B',BizDayConvention.Preceding)
    # spotBeginDate = Date(2006, 1, 1)
    begin = spotBeginDate.toISO().replace('-', '')
    end = endDate.toISO().replace('-', '')
    
    fields = ['tradeDate', 'openPrice', 'highestPrice', 'lowestPrice', 'closePrice']
    security = DataAPI.MktFunddGet(secID=secID, beginDate=begin, endDate=end, field=fields)
    security['amp1'] = security['highestPrice']/security['closePrice']
    security['amp2'] = security['highestPrice']/security['openPrice']    
    security['amp3'] = security['lowestPrice']/security['closePrice']
    security['amp4'] = security['lowestPrice']/security['openPrice']    
    security['u'] = np.log(security['amp1'])*np.log(security['amp2']) + np.log(security['amp3'])*np.log(security['amp4'])
    security['tradeDate'] = pd.to_datetime(security['tradeDate'])
    
    periods = {'hv1W': 5, 'hv2W': 10, 'hv1M': 21, 'hv2M': 41, 'hv3M': 62, 'hv4M': 83, 
               'hv5M': 104, 'hv6M': 124, 'hv9M': 186, 'hv1Y': 249, 'hv2Y': 497}
    # 利用Parkinson模型计算波动率
    for prd in periods.keys():
        tmp = np.sqrt(pd.rolling_mean(security['u'], window=periods[prd]))*math.sqrt(252.0)
        security[prd] = np.round(tmp, 5)
    
    security = security[security.tradeDate >= beginDate.toISO()]
    security = security.set_index('tradeDate')
    return security

## 计算一段时间标的的Garman Klass历史波动率，返回值包括以下不同周期的波动率：
# 一周，半月，一个月，两个月，三个月，四个月，五个月，半年，九个月，一年，两年
def getHistVolatilityGarmanKlass(secID, beginDate, endDate):
    cal = Calendar('China.SSE')
    spotBeginDate = cal.advanceDate(beginDate,'-520B',BizDayConvention.Preceding)
    # spotBeginDate = Date(2006, 1, 1)
    begin = spotBeginDate.toISO().replace('-', '')
    end = endDate.toISO().replace('-', '')
    
    fields = ['tradeDate', 'openPrice', 'highestPrice', 'lowestPrice', 'closePrice', 'preClosePrice']
    security = DataAPI.MktFunddGet(secID=secID, beginDate=begin, endDate=end, field=fields)
    
    # O
    security['sigmaO'] = np.power(np.log(security['openPrice']/security['preClosePrice']), 2)
    # C
    security['sigmaC'] = np.power(np.log(security['closePrice']/security['openPrice']), 2)
    # P
    security['sigmaP'] = np.power(np.log(security['highestPrice']/security['lowestPrice']), 2)/4.0/np.log(2.0)
    # RS
    security['amp1'] = security['highestPrice']/security['closePrice']
    security['amp2'] = security['highestPrice']/security['openPrice']    
    security['amp3'] = security['lowestPrice']/security['closePrice']
    security['amp4'] = security['lowestPrice']/security['openPrice']    
    security['sigmaRS'] = np.log(security['amp1'])*np.log(security['amp2']) + np.log(security['amp3'])*np.log(security['amp4'])
    
    security['tradeDate'] = pd.to_datetime(security['tradeDate'])
    periods = {'hv1W': 5, 'hv2W': 10, 'hv1M': 21, 'hv2M': 41, 'hv3M': 62, 'hv4M': 83, 
               'hv5M': 104, 'hv6M': 124, 'hv9M': 186, 'hv1Y': 249, 'hv2Y': 497}
    # 利用Parkinson模型计算波动率
    for prd in periods.keys():
        sigmaO2 = pd.rolling_mean(security['sigmaO'], window=periods[prd])
        sigmaC2 = pd.rolling_mean(security['sigmaC'], window=periods[prd])
        sigmaP2 = pd.rolling_mean(security['sigmaP'], window=periods[prd])
        sigmaRS2 = pd.rolling_mean(security['sigmaRS'], window=periods[prd])
        sigma2 = np.sqrt(sigmaO2 - 0.383*sigmaC2 + 1.364*sigmaP2 + 0.019*sigmaRS2)*math.sqrt(252.0)
        security[prd] = np.round(sigma2, 5)
    
    security = security[security.tradeDate >= beginDate.toISO()]
    security = security.set_index('tradeDate')
    return security

## 计算一段时间标的的Yang Zhang历史波动率，返回值包括以下不同周期的波动率：
# 一周，半月，一个月，两个月，三个月，四个月，五个月，半年，九个月，一年，两年
def getHistVolatilityYangZhang(secID, beginDate, endDate):
    cal = Calendar('China.SSE')
    spotBeginDate = cal.advanceDate(beginDate,'-520B',BizDayConvention.Preceding)
    # spotBeginDate = Date(2006, 1, 1)
    begin = spotBeginDate.toISO().replace('-', '')
    end = endDate.toISO().replace('-', '')
    
    fields = ['tradeDate', 'openPrice', 'highestPrice', 'lowestPrice', 'closePrice', 'preClosePrice']
    security = DataAPI.MktFunddGet(secID=secID, beginDate=begin, endDate=end, field=fields)
    
    # O
    security['sigmaO'] = np.log(security['openPrice']/security['preClosePrice'])
    # C
    security['sigmaC'] = np.log(security['closePrice']/security['openPrice'])
    # RS
    security['amp1'] = security['highestPrice']/security['closePrice']
    security['amp2'] = security['highestPrice']/security['openPrice']    
    security['amp3'] = security['lowestPrice']/security['closePrice']
    security['amp4'] = security['lowestPrice']/security['openPrice']    
    security['sigmaRS'] = np.log(security['amp1'])*np.log(security['amp2']) + np.log(security['amp3'])*np.log(security['amp4'])
    
    security['tradeDate'] = pd.to_datetime(security['tradeDate'])
    periods = {'hv1W': 5, 'hv2W': 10, 'hv1M': 21, 'hv2M': 41, 'hv3M': 62, 'hv4M': 83, 
               'hv5M': 104, 'hv6M': 124, 'hv9M': 186, 'hv1Y': 249, 'hv2Y': 497}
    # 利用Parkinson模型计算波动率
    for prd in periods.keys():
        sigmaO2 = np.power(pd.rolling_std(security['sigmaO'], window=periods[prd]), 2)
        sigmaC2 = np.power(pd.rolling_std(security['sigmaC'], window=periods[prd]), 2)
        sigmaRS2 = pd.rolling_mean(security['sigmaRS'], window=periods[prd])
        k = 0.34/(1.34+(periods[prd]+1)/(periods[prd]-1))
        sigma2 = np.sqrt(sigmaO2 + k*sigmaC2 + (1-k)*sigmaRS2)*math.sqrt(252.0)
        security[prd] = np.round(sigma2, 5)
    
    security = security[security.tradeDate >= beginDate.toISO()]
    security = security.set_index('tradeDate')
    return security
```
