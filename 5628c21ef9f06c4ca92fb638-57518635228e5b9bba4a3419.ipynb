# 分钟线回测性能测试

> 本例试图分析当前分钟线回测性能以及可能的瓶颈。

0. 测试原理
---------------
***

运行两个配置一样的策略，区别：

* 第一个策略不做任何操作，可以认为这个策略主要的消耗在IO上；
* 第二个策略具有实际的计算功能，可以认为这个策略的运行时间约等于第一个策略的时间+策略逻辑计算时间。

1. 测试对比（时间1年，股票1支）
-----------------
***

本测试中的结果，单一策略：

* IO时间：约等于25秒；

* 计算时间：约等于5秒。

2. 测试对比（时间2年，股票5支）
-----------------------------------
***

本测试中的结果，单一策略：

* IO时间：约等于1分15秒；

* 计算时间：约等于1分10秒。

3. . 测试对比（时间2年，股票20支）
-----------------------------------
***

本测试中的结果，单一策略：

* IO时间：约等于2分22秒；

* 计算时间：约等于6分22秒。

3. 总结
-----------------
***

可以看到:

* IO损耗的增长显著低于线性水平
* 计算逻辑的消耗增长可能约等于或高于线性水平

随着时间的加长以及股票池的扩大，计算逻辑的耗时增长远远大于IO消耗的增长。再超过一定阈值之后，这里测试的结果（约为2年，5支股票），计算逻辑的耗时会成为主要的耗时部分。
