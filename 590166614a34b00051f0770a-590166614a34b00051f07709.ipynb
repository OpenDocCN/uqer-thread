# 事件驱动策略研究2——员工持股计划，近年来alpha最高的事件？

类似本人上一篇《机构调研（添加止盈止损）》的分析框架，对“员工持股计划”这一近年来新出现的事件进行分析，事件信息同样采自Wind数据库，数据清洗整理后存为“yuangongchigu_new.csv”

员工持股计划概述：
2014年6月，证监会发布《关于上市公司实施员工持股计划试点的指导意见》，我国上市公司员工持股计划正式拉开序幕，截至2017年3月，已经有了500多家上市公司发布了员工持股计划董事会预案，作为一种新型的事件投资机会，吸引着广大投资者的关注。
它的投资逻辑主要有以下几点：
1. 激励员工。员工持股计划将员工的收益与股价挂钩，激励员工努力工作，提高公司的业绩。
2. 增加上市公司股价诉求。增加了公司对于股价的诉求，上市公司的市值管理也会更加地积极。
3. 安全垫。员工持股计划的实施价无形中给股价提供了一个支撑位，为二级市场的投资者提供了一个安全垫。
4. 增强对公司未来的乐观预期。买股票其实就是买预期，员工持股计划能够提供一个乐观的预期，在这种预期的持续发酵中，能够大概率地跑赢市场。
5. 新增买入力量。除认购非公开发行股票和股东赠与方式的员工持股计划以外，其他方式的员工持股计划需要在二级市场买入股票，为市场带来了额外的买盘，可视为重要的上涨推动力量。
综上，员工持股计划有利于提高员工的工作积极性,长期来看对公司发展具有正面影响。同时,作为目前国企混改的主要方式之一,员工持股计划已成为众多投资者所关注的事件,不少研报将其称为近些年alpha最高的事件。


```python
事件策略回测
以2014年至今的所有员工持股计划事件为全样本（优选其中员工持股比例大于70%的作为优选样本，大概380多次预案），进行策略回测
回测逻辑：首先获取员工持股计划事件信息，在预案后i日入场，然后持股j天，如果中间上市公司发布了实施成功公告，则在公告后5日卖出（这里选择5日后卖出，是因为有研报统计发现公告成功对股价还有利好），从2014年8月份第一次员工持股事件回测到2017年3月份，观察事件是否真有超额收益
```

```python
import seaborn
from matplotlib import pylab
import matplotlib.pyplot as plt
from matplotlib.pyplot import *
from CAL.PyCAL import *    # CAL.PyCAL中包含font
from datetime import timedelta,datetime
import pandas as pd
from string import zfill
import quartz as qz
raw_df=pd.read_csv('yuangongchigu_new.csv')
raw_df['ticker'] = raw_df['ticker'].apply(str).apply(lambda x: zfill(x, 6))
def get_specific_trading_day(date_str='2016-01-04', window=20, direction='back'):
    """给定日期，可以通过这个函数获取前后N天的那个交易日"""
    date = datetime.strptime(date_str, '%Y-%m-%d')
    if direction=='back':
        begin_day = date - (window+5) * 4 * timedelta(days=1)   # 可以获取冗余的交易日，防止最终取不到有效数据
        end_day = date
    elif direction=='forward':
        begin_day = date
        end_day = date + (window+5)  * 4 * timedelta(days=1) 
        
    begin_day_str = begin_day.strftime('%Y-%m-%d')
    end_day_str = end_day.strftime('%Y-%m-%d')
        
    cal_dates = DataAPI.TradeCalGet(exchangeCD=u"XSHG", beginDate=begin_day_str, endDate=end_day_str, field="calendarDate,isOpen")
    trading_days = cal_dates[cal_dates['isOpen']==1]["calendarDate"].tolist()
    # trading_days = [day[0:4] + day[5:7] + day[8:] for day in trading_days]  # 更改日期的格式，将'2014-01-03'转化为'20140103'
    try:
        if direction=='back':
            return trading_days[-window]
        return trading_days[window]
    except:
        print date_str,window,direction
raw_df['t_plus_1'] = raw_df['date'].apply(lambda x: get_specific_trading_day(x, 5,'forward'))

raw_df['secID'] =raw_df['ticker'].map(lambda x: x + '.XSHG' if x[0] == '6' else x + '.XSHE')
raw_df_new=raw_df['date_2'].dropna()
raw_df['t_plus_2']=np.zeros(len(raw_df))
for i in range(len(raw_df_new)):
   
      raw_df['t_plus_2'][i] = get_specific_trading_day(raw_df['date_2'][i],5,'forward')
```

全样本事件回测，观察是否有超额收益  （i,j）分别表示预案后几日进场和最大持股时间

```python
def backtest_time(i, j):  ##i表示预案后几日进场，j表示最大持股时间
    raw_df['t_plus_1'] = raw_df['date'].apply(lambda x: get_specific_trading_day(x,i,'forward'))
    start = '2014-08-27'                       # 回测起始时间,事件首个交易日
    end = '2017-03-22'                         # 回测结束时间
    benchmark = 'ZZ500'                        # 策略参考标准,中证500
    universe = set_universe('A')  # 证券池，支持股票和基金
    capital_base = 10000000                     # 起始资金
    freq = 'd'                                 # 策略类型，'d'表示日间策略使用日线回测，'m'表示日内策略使用分钟线回测
    refresh_rate = 1                           # 调仓频率，表示执行handle_data的时间间隔，若freq = 'd'时间间隔的单位为交易日，若freq = 'm'时间间隔为分钟
    commission = Commission(buycost=0.0000, sellcost=0.0000)

    def initialize(account):                   # 初始化虚拟账户状态
        account.day_counter = {}  #用于记录持仓天数

    def handle_data(account):                  # 每个交易日的买入卖出指令
    
        for stk in account.avail_security_position:  # 记录持仓的天数
            if account.day_counter.has_key(stk):
                     account.day_counter[stk] += 1
            else:
                     account.day_counter[stk] = 1
      
            
   ##卖出，最大到j日 卖出         
        for stk in account.security_position:
            if(account.day_counter[stk]>j):   #持仓超过j天，卖出
                     order_to(stk,0)
        for stk in (set(account.day_counter.keys())-set(account.security_position.keys())):
                     del account.day_counter[stk]
    ##   卖出，持有至实施完成后第5个交易日卖出     
        univ_df_sell=raw_df[raw_df['t_plus_2']==account.current_date.strftime('%Y-%m-%d')]
        if len(univ_df_sell)>0: ##当天有员工持股事件到期
            univ_sell=univ_df_sell['secID'].tolist()
            univ_sell=list(set(univ_sell).intersection(set(account.security_position)))###求当天有股票持有 ，但是未到35日就到期
            for stk in univ_sell:
                    order_to(stk, 0)
        for stk in (set(account.day_counter.keys())-set(account.security_position.keys())):
            del account.day_counter[stk]
    ##判断当前持股数量        
        current_stock_number=len(account.avail_security_position.keys())
    
    ##买进
        univ_df = raw_df[raw_df['t_plus_1']==account.current_date.strftime('%Y-%m-%d')]
     
        max_number=10000  ###表示不限仓
    
    ##如果当日股票持仓数量少于max_number个，则增仓买入新股票
        if current_stock_number<max_number:
            if len(univ_df )>0: # 当天有事件发生
                need_buy_number=max_number-current_stock_number
                univ_df=univ_df.tail(need_buy_number)
                univ = univ_df ['secID'].tolist()      
                univ = list(set(univ) - set(account.security_position))  # 已经持有的股票不再重新下单
                need_buy_stock_number=len(univ)##需要新进买入股票的数量
                number=current_stock_number+need_buy_stock_number
            
            
                for stk in univ:
                    order_pct_to(stk,1.0/number)
                for stk in account.security_position:
                    order_pct_to(stk,1.0/number)

       
    bt, perf = qz.backtest(universe=universe, start=start, end=end,
                                initialize=initialize, handle_data=handle_data,
                                capital_base=capital_base, refresh_rate=refresh_rate,
                                freq=freq)
    return perf
```

```python
def backtest(i, j):  ##i表示止盈率，j表示止损率
    start = '2014-08-27'                       # 回测起始时间,事件首个交易日
    end = '2017-03-22'                         # 回测结束时间
    benchmark = 'ZZ500'                        # 策略参考标准,中证500
    universe = set_universe('A')  # 证券池，支持股票和基金
    capital_base = 10000000                     # 起始资金
    freq = 'd'                                 # 策略类型，'d'表示日间策略使用日线回测，'m'表示日内策略使用分钟线回测
    refresh_rate = 1                           # 调仓频率，表示执行handle_data的时间间隔，若freq = 'd'时间间隔的单位为交易日，若freq = 'm'时间间隔为分钟
    commission = Commission(buycost=0.0002, sellcost=0.0002)

    def initialize(account):                   # 初始化虚拟账户状态
        account.day_counter = {}  #用于记录持仓天数

    def handle_data(account):                  # 每个交易日的买入卖出指令
    
        for stk in account.avail_security_position:  # 记录持仓的天数
            if account.day_counter.has_key(stk):
                     account.day_counter[stk] += 1
            else:
                     account.day_counter[stk] = 1
      ##止盈止损
        for stk in account.avail_security_position:
            if (account.reference_price[stk])/account.security_cost[stk]>=i or (account.reference_price[stk])/account.security_cost[stk]<=j:
                     order_to(stk,0)
        for stk in (set(account.day_counter.keys())-set(account.security_position.keys())):
                     del account.day_counter[stk]  
            
   ##卖出，最大到35日 卖出         
        for stk in account.security_position:
            if(account.day_counter[stk]>35):   #持仓超过35天，卖出
                     order_to(stk,0)
        for stk in (set(account.day_counter.keys())-set(account.security_position.keys())):
                     del account.day_counter[stk]
    ##   卖出，持有至实施完成后第5个交易日卖出     
        univ_df_sell=raw_df[raw_df['t_plus_2']==account.current_date.strftime('%Y-%m-%d')]
        if len(univ_df_sell)>0: ##当天有员工持股事件到期
            univ_sell=univ_df_sell['secID'].tolist()
            univ_sell=list(set(univ_sell).intersection(set(account.security_position)))###求当天有股票持有 ，但是未到35日就到期
            for stk in univ_sell:
                    order_to(stk, 0)
        for stk in (set(account.day_counter.keys())-set(account.security_position.keys())):
            del account.day_counter[stk]
    ##判断当前持股数量        
        current_stock_number=len(account.avail_security_position.keys())
    
    ##买进
        univ_df = raw_df[(raw_df['t_plus_1']==account.current_date.strftime('%Y-%m-%d'))]
     
        max_number=10000
    
    ##如果当日股票持仓数量少于max_number个，则增仓买入新股票
        if current_stock_number<max_number:
            if len(univ_df )>0: # 当天有高管增持事件发生
                need_buy_number=max_number-current_stock_number
                univ_df=univ_df.tail(need_buy_number)
                univ = univ_df ['secID'].tolist()      
                univ = list(set(univ) - set(account.security_position))  # 已经持有的股票不再重新下单
                need_buy_stock_number=len(univ)##需要新进买入股票的数量
                number=current_stock_number+need_buy_stock_number
            
            
                for stk in univ:
                    order_pct_to(stk,1.0/number)
                for stk in account.security_position:
                    order_pct_to(stk,1.0/number)

       
    bt, perf = qz.backtest(universe=universe, start=start, end=end,
                                initialize=initialize, handle_data=handle_data,
                                capital_base=capital_base, refresh_rate=refresh_rate,
                                freq=freq)
    return perf
```

下面考虑限仓（实盘过程小资金不限仓不太实际，这里选择限仓为10股），策略逻辑为只有当当日有空余仓位时，而且当日前10日到当日前5日这段时间段内有员工持股计划事件公告时，才入仓买入该股(优选时间靠后的)，最大买到持股上限（10股），每个调仓日从新分配股票池内所有股票权重，使之等权。如果股票池内的股票达到止盈或止损点，则卖出(策略优化阶段设置费率为万四，滑点为0)
