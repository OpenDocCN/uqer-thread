复杂事件处理及其策略
=========================

在进行量化策略设计和执行时，主流的做法是基于事件驱动的数据流处理架构。

为了对复杂事件处理有深入的理解，我会以《朝闻道 CEP》系列的形式，逐步探究一些常见的复杂事件。


事件可以分为简单事件和复杂(复合)事件，复杂事件是源于简单事件。

从而，我们可以将整个架构理解成如下三个层次。
> * 最底层，是数据流，数据流中的内容可以是时间、收盘价、成交量、order book上的买一价、新闻文本等任何形式的数据。 
> * 数据流之上，是事件流，事件流中的内容可以是某一证券的最新Tick价格，可以是通过新闻文本处理过的新闻事件等时序性的，进行过包装的简单事件
> * 事件流之上，是复杂事件流，是对简单事件的进一步处理。复杂事件可以是某个技术指标出现上穿下穿，order book上的委托单数量总和，order book上的买卖价差等将简单事件进一步处理，或者将多个简单事件进行组合而成的复杂事件

#### 事件的关系
事件的关系决定了复杂事件的种类。事件的关系主要有5种[1] ：
>1. 时间顺序关系
>
>	动作事件和动作事件之间，动作事件和状态变化事件之间，都存在时间顺序。
>	
>	比如成交量逐步缩小，体现了成交量序列的时序关系。
>2. 聚合关系
>
>	动作事件和动作事件之间，状态事件和状态事件之间都存在聚合关系。即个体的聚合形成整体集合。

>	比如当天所有交易的VWAP值，是对当天的价格和交易量的聚合。
>3. 层次关系
>
>	动作事件和动作事件之间，状态事件和状态事件之间都存在层次关系，即父类事件和子类事件的层次关系，从父类到子类是具体化，从子类到父类是泛化。
>4. 依赖关系
>
>	事物的状态属性之间彼此的依赖关系和约束关系。
>5. 因果关系
>
>	对于完整的动作过程，结果状态为果，初始状态和动作都可以视为原因。类比哲学上论述事物如何发展也是有两个因素的，一是内部本质，二是外部作用.
