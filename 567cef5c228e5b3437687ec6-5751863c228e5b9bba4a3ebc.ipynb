这两件事情都是紧密围绕着大赛的策略要求来的，所以我们得详细地考察下策略要求：

**策略要求**

优矿将在每一期结束后校验获奖策略是否满足参赛要求，如果不满足要求，优矿有权取消该策略的名次。参赛策略原则上满足如下要求：

1. 只允许投资于沪深两市上市的A股股票；
2. 不允许投资于S、ST、\*ST、S\*ST以及SST类上市公司股票；
3. 不允许投资于流动性过差的股票，包括在每个交易日之前的20个交易日平均日交易量小于1000万人民币的股票（如果股票在过去20个交易日或停牌或涨跌停，通联数据将自行决定是否禁买）；
4. 单只股票持股不得高于10%；
5. 股票仓位在任意三个以上收盘日低于80%则不达标；
6. 相对HS300，强制平仓线90%。
7. 首日上市新股（或复牌首日股票）等当日不设涨跌停板限制的股票；
8. 参赛策略不允许在参赛期间进行手动调仓。

分类来看，这8条要求可以大概分成如下几类：
(1) 可买股票限制，包括1、2、3、7四条
(2) 持仓限制，包括4、5两条
(3) 收益限制，即第6条
(4) 其他，即第8条

**检查bt和perf**
-----------------------

饭要一口一口吃，事要一件一件做。先来看第一件事情吧。

第(4)条不会有任何影响；bt和perf中已经有了每天的持仓和收益数据，所以(2)和(3)很好检查；麻烦的是(1)，可能需要利用DataAPI取一取数据，不过按照要求一条一条取就是，难倒不难。

所以可以写出下面的代码，其中的check函数接受bt和perf两个参数，可以对策略的结果进行分析：

```python
from datetime import datetime

import numpy as np
import pandas as pd

from CAL.PyCAL import *


def normalize_date(date):
    """
    处理日期
    """
    
    date = pd.tseries.tools.normalize_date(pd.Timestamp(date))
    return datetime(date.year, date.month, date.day)

def validate(universe):
    """
    去除非A股股票
    """
    
    return [sec for sec in universe if sec[0] in '36' or sec[:1] == '00']

def getdata(start, end, universe, benchmark='000300.ZICN', batchsize=5000):
    """
    取数据
    """
    
    cal = Calendar('China.SSE')
    start = normalize_date(start)
    end = normalize_date(end)
    s1 = cal.advanceDate(start, '-20B', BizDayConvention.Following)
    s1 = datetime(s1.year(), s1.month(), s1.dayOfMonth())
    
    # 以防策略的benchmark不是沪深300
    df_bm = DataAPI.MktIdxdGet(indexID = benchmark, beginDate = start.strftime('%Y%m%d'), endDate = end.strftime('%Y%m%d'), field = ['tradeDate', 'closeIndex', 'preCloseIndex'])
    ret = (df_bm['closeIndex'] / df_bm['preCloseIndex'] - 1.).tolist()
    bm_cret = [ret[0]]
    for r in ret[1:]:
        bm_cret.append((1 + bm_cret[-1]) * (1 + r) - 1)

    S = len(universe)
    N = batchsize / (end - start).days
    if N <= 1:
        batches = [[stock] for stock in universe]
    elif N > S:
        batches = [universe]
    else:
        batches = [universe[i:min(i+N,S)] for i in range(0, S, N)]
    
    # 平均成交量与停复牌信息
    df_sc = {}
    for b in batches:
        df_b = DataAPI.MktEqudAdjGet(secID=b, beginDate=s1.strftime('%Y%m%d'), endDate=end.strftime('%Y%m%d'), field=['secID', 'tradeDate', 'openPrice', 'closePrice', 'turnoverValue'])
        for sec in b:
            df = df_b[df_b.secID == sec].loc[:, ['tradeDate', 'openPrice', 'closePrice', 'turnoverValue']]
            df['openPrice'] = df['openPrice'].shift(1)
            df['mtv'] = [np.nan] + pd.rolling_mean(df.turnoverValue, 20).tolist()[:-1]
            df_sc[sec] = df[df.tradeDate >= start.strftime('%Y-%m-%d')].loc[:, ['tradeDate', 'openPrice', 'closePrice', 'mtv']]
    
    # ST及其他
    df_st0 = {}
    for b in batches:
        try:
            df_b = DataAPI.SecSTGet(secID=b, beginDate=start.strftime('%Y%m%d'), endDate=end.strftime('%Y%m%d'), field=['tradeDate', 'secID'])
            for sec in b:
                df_st0[sec] = df_b[df_b.secID == sec]
        except:
            pass
    df_st = {'tradeDate':[], 'stsecs': []}
    for date in df_bm['tradeDate']:
        df_st['tradeDate'].append(date)
        df_st['stsecs'].append([])
        for sec in df_st0:
            df = df_st0[sec]
            if len(df) and len(df[df.tradeDate == date]):
                df_st['stsecs'][-1].append(sec)
    df_st = pd.DataFrame(df_st)
    
    return bm_cret, df_sc, df_st

def check(bt, perf):
    """
    检查策略合格情况
    """
    
    trading_days = bt.tradeDate.tolist()
    start = normalize_date(trading_days[0])
    end = normalize_date(trading_days[-1])
    
    universe = set()
    for pos in bt.security_position:
        universe = universe.union(pos.keys())
    universe = validate(universe)
    
    bm_cret, df_sc, df_st = getdata(start, end, universe)
    pt_cret = perf['cumulative_returns']
    
    for i,date in enumerate(trading_days):
        if pt_cret.iat[i] - bm_cret[i] < -0.1:
            print '{} 累计收益低于沪深300达10%，会被强制平仓'.format(dstr)
        
        dstr = date.strftime('%Y-%m-%d')
        row = bt.iloc[i]
        cash, ptvalue = row['cash'], row['portfolio_value']
        
        if cash / ptvalue >= 0.2:
            print '{} 股票持仓不满80%'.format(dstr)
            
        secvalue =  - row['cash']
        for sec,amt in row['security_position'].items():
            if sec[0] == '5':
                print '{} 非A股上市股票: {}'.format(dstr, sec) 
            elif sec[0] == '0' and sec[1] != '0':
                print '{} 非A股上市股票: {}'.format(dstr, sec)
                
            if sec in df_st[df_st.tradeDate == dstr]['stsecs'].iat[0]:
                print '{} ST或其他股票: {}'.format(dstr, sec)
            
            if sec in df_sc:
                df = df_sc[sec]
                df = df[df.tradeDate == dstr]
                opnp = df['openPrice'].iat[0]
                if np.isnan(opnp) or opnp == 0:
                    print '{} 新上市或复牌股票: {}'.format(dstr, sec)

                clsp = df['closePrice'].iat[0]
                if clsp * amt['amount'] / ptvalue > 0.1:
                    print '{} 单只股票持仓超过10%: {}'.format(dstr, sec)
                    
                mtv = df['mtv'].iat[0]
                if mtv < 10**7:
                    print '{} 流动性过差股票: {}'.format(dstr, sec)
```
