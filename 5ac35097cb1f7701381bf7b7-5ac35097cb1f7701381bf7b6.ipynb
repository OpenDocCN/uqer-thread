## 1. 资产配置理论的发展历程

###  1.1 Mean-Variance Model
关于资产配置的理论，最早可追溯到Markowitz（1953）的均值-方差模型框架下的投资组合理论。该模型假设理性投资者通过寻求最优资产配置比例，实现既定组合风险水平条件下的期望收益最大化的目标。<br>

### 1.2 Equal-Weight Model and Lowest Volatility Model
均值-方差模型存在以下不足之处：
1. 模型对于参数选取过于敏感，对模型参数进行微调，可能得到的资产配置比例与此前相去甚远；
2. 模型使用的是组合期望收益和期望风险，而使用历史值对未来进行预测往往不够准确；
3. 由于模型考虑的是组合的整体风险，而忽视了组合中风险的构成，在构建投资组合时，往往使得风险在某类资产上过于集中。

为了解决上述问题，Evans（1968）等人从投资组合中资产数量的角度衡量风险分散程度，主要方法是构建**等权重组合和最小方差组合**。<br>
等权重组合是将各个资产的权重平均分配，若各类资产具有相同的均值、方差、协差阵，则等权重组合位于均值-方差模型的有效前沿上；最小方差组合是均值-方差模型中有效前沿上一个存在特殊解的点，由于求解这个特殊点不需要使用预期收益的信息，因而构建出的组合被认为是稳健的。 <br>

### 1.3 Risk Parity Model

#### 1.3.1 风险平价模型的概念

由于基于传统投资组合理论构建的等权重组合及方差最小组合并不能真正地分散组合风险，磐安基金（PanAgora）的首席投资官钱恩平（Edward Qian）提出了风险平价策略，并被桥水基金（Bridge Water）运用于实际投资中，获得了巨大成功。桥水基金旗下的全天候基金（All Weather）即为基于风险平价策略的基金，在1996至2015年，该基金年化收益超过同期标准普尔500指数3.07%，而波动率比同期标普500低6.74%，夏普比率达到了0.64。<br>
风险平价策略提倡配置风险而不是配置资产。传统的投资组合理论强调控制整体的绝对风险，即整个组合的波动性；而风险平价策略强调控制各类资产的相对风险，即使得各类资产的风险处于相对均衡的水平。由于组合的风险达到了平衡，故理论上而言，通过风险平价策略进行资产配置可以抵御各类风险事件，此即为桥水基金全天候投资策略的基础。<br>
根据风险平价策略的思想，可通过推导风险平价模型中风险贡献（Risk Contributioin，RC）的计算公式，利用达到风险平价时组合中各类资产风险贡献相等的条件，求解风险平价模型中各类资产的最优权重。<br>

#### 1.3.2 风险平价模型表达式
假设资产组合的权重向量为$\omega$，组合协差阵为$\Sigma$，组合标准差为$\sigma$。则定义风险贡献为：
$RC_i = \omega_i\frac{\partial\sigma}{\partial\omega_i} = \frac{\omega_i(\Sigma\omega)_i}{\sqrt{\omega^\mathsf{T}\Sigma\omega}}$   

其中，组合标准差与权重向量和组合协差阵的关系为:$\sigma = \sqrt{\omega^\mathsf{T}\Sigma\omega}$  
一般情况下，风险平价模型采用序列二次规划算法（Sequetial Quadratic Programming，SQP）予以求解。

### 2.3 最小方差模型回测
最小方差模型的核心是，计算出最小方差组合中各个证券的权重。

根据M-V模型，最小方差模型应满足：
$  min (\sigma_p) = \sum_i\sum_j{w_i w_j Cov(\sigma_i,\sigma_j)} $
其中，$s.t. \sum_i{w_i} = 1$。

在Python中实现最小方差模型的关键是利用约束优化算法（SLSQP）求得最小值，该算法在scipy.optimize中的minimize函数中使用。
> scipy.optimize.minimize(func,x0,constrains,bounds,method)使用方法：
首先设置初始值、约束条件函数和边界值。
这里，初始值可以设置为$\frac{1}{N}$，每个权重值的边界均设置为(0,1).

## 3. 总结
通过上述简单回测可以发现：
1. 等权重组合收益较高，回撤较大；组合收益对权益市场的依赖度较高
2. 最小方差组合能够较大程度地规避对权益市场分配过高权重，从而降低投资风险，但问题也显而易见，即该组合收益较低
3. 风险平价组合较好地分配了投资组合的风险度，同时收益好于最小方差组合
