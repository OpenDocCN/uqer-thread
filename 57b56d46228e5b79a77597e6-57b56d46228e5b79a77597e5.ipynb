```python
import pandas
def flowAggByInd(date, universe):
    flow = DataAPI.MktEquFlowOrderGet(secID=universe,ticker=u"",beginDate=date,endDate=date,field=u"",pandas="1")
    negMarketValue = DataAPI.MktEqudAdjAfGet(secID = universe , isOpen="",tradeDate=date,field=['secID','negMarketValue'],pandas="1")
    flow = flow.merge(negMarketValue, left_on = "secID", right_on = 'secID')
    ind = DataAPI.EquIndustryGet(industryVersionCD=u"010303",industry=u"",secID=universe,ticker=u"",intoDate=date,field=[u"secID",u"industryID1", u"industryName1"],pandas="1")
    groups = flow.merge(ind, left_on='secID', right_on="secID").groupby(u'industryName1')
    res = pandas.DataFrame()
    for k in groups.groups.keys()[0:]:
        g = groups.get_group(k)
        d = g[["inflowS","inflowM","inflowL","inflowXl","outflowS","outflowM","outflowL","outflowXl","netInflowS","netInflowM","netInflowL","netInflowXl", "negMarketValue"]].sum()
        for c in ["inflowS","inflowM","inflowL","inflowXl","outflowS","outflowM","outflowL","outflowXl","netInflowS","netInflowM","netInflowL","netInflowXl"]:
            d[c] = d[c]/d["negMarketValue"]
        d['date'] = g.tradeDate.values[-1]
        d['industryName'] = g.industryName1.values[-1]
        res = res.append(d, ignore_index=True)
    return res
def flowAggByIndustryAndWeeks(d1, d2, universe): # e.g. d1 ='20160801', d2 = '20160818'
    #trade date list from d1 to d2
    datelist = DataAPI.TradeCalGet(exchangeCD=u"XSHG,XSHE",beginDate = d1, endDate = d2).drop_duplicates('calendarDate')
    #only open
    datelist = datelist[datelist.isOpen == 1]
    # assume start week is 1, then generate week list
    weekNum = 1
    weekList = []
    for i in range(len(datelist)):
        isweekend = datelist.iloc[i]['isWeekEnd']
        if isweekend == 0:
            weekList.append(weekNum)
        else:
            weekList.append(weekNum)
            weekNum = weekNum + 1    
    datelist['weekNum'] = weekList 
    #groupby week number
    group = datelist.groupby('weekNum')
    
    resOverWeek = pandas.DataFrame()
    #loop over all weeks
    for k in group.groups.keys()[0:]:
        #print 'week',k
        g = group.get_group(k)
        #combine data by individual date in the week
        sublist = g.calendarDate.tolist()
        weekRes = pandas.DataFrame()
        for d in sublist:
            res = flowAggByInd(d, universe)
            weekRes = weekRes.append(res)
        # groupby industry to aggregate over the week
        weekGroups = weekRes.groupby('industryName')
        #loopover all industry
        for subk in weekGroups.groups.keys():
            #print 'subk', subk
            subres = weekGroups.get_group(subk)
            # sum over the week for each industry
            d = subres[["inflowS","inflowM","inflowL","inflowXl","outflowS","outflowM","outflowL","outflowXl","netInflowS","netInflowM","netInflowL","netInflowXl"]].sum()
            #use the last date of the week as date
            d['date'] = sublist[-1]
            d['industryName'] = subk
            resOverWeek = resOverWeek.append(d,ignore_index=True)
    return resOverWeek
```
