### 3. 预案公告到最后实施，会有多少个交易日？

```python
tmp = df #[['预案公告日', '除权除息日']]
tmp['预案公告日'] = tmp['预案公告日'].apply(lambda x: dt.datetime.strptime(str(x)[:10], '%Y-%m-%d'))
tmp['除权除息日'] = tmp['除权除息日'].apply(lambda x: dt.datetime.strptime(x, '%Y-%m-%d'))
tmp['间隔天数'] = (tmp['除权除息日'] - tmp['预案公告日']).apply(lambda x: x.days)
```

```python
pd.DataFrame(tmp['间隔天数'].describe())
```

### 7. 预案公告日到股权登记日期间股价涨跌幅？

```python
tmp = df
tmp = tmp[(tmp['股权登记日收盘价'] > 0) & (tmp['预案公告当日开盘价'] > 0)]
tmp['涨跌幅'] = (tmp['股权登记日收盘价'] / tmp['预案公告当日开盘价'] - 1).apply(lambda x: round(x, 4))
tmp[['涨跌幅']].head()
```

```python
tmp[['涨跌幅']].describe()
```

```python
number, edge = np.histogram(tmp[['涨跌幅']], bins=20)
change_dist = pd.DataFrame({'number': number, 'edge': edge[1: ]})
change_dist
```
