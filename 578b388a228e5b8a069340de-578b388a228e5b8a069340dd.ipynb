发现回测HS300时使用set_universe("HS300") 会引起年化14%左右的误差

对这个误差的原因猜想有几种：

1. 幸存者偏差
2. 权重， HS300的每个股票的权重不一样， 回测用的是等权。 （个人觉得等权不会造成14%的年化收益的偏差）
3. 剔除的停牌， ST， 流通市值的影响。 （这个影响也不应该造成这么大的偏差）
4. 红除权对指数的影响。 （每年1%够了吧？）

经过测试之后，发现由#1. 幸存者偏差引起的年化收益误差在36%。 （前一版回测代码筛选股票处发生错误导致回测结果不准确，下面是修正后的回测）

改良方法：
1. 不使用set_universe函数来设置universe变量。
2. 自己重写get_universe方法，

所有被编入过HS300指数的股票集合，包括最新的HS300成分股和中间被剔除掉的股票成分股
3. 买卖时用IdxCloseWeightGet或IdxConsGet获取买卖当天的HS300指数的成分股列表进行操作。
