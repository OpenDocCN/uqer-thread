# 计算所有品种iv

```python
import pandas as pd
from CAL.PyCAL import *
import datetime

def GetAllOPOneDay(date):
    '''获得某一天的所有期权'''
    OPsOneDay = DataAPI.MktOptdGet(secID=u"",optID=u"",ticker=u"",tradeDate=date,
                                          exchangeCD=u"",field=[u'secID',u'ticker',u'secShortName',
                                                                u'tradeDate',u'closePrice',
                                                                u'turnoverVol',u'varTicker'],pandas="1")
    '''筛掉没有成交量的合约'''
    OPsOneDay = OPsOneDay[OPsOneDay.turnoverVol>3] 
    '''添加字段'''
    OPsOneDay[u"s"] = 0
    OPsOneDay[u"biaodiwu"] = OPsOneDay.secShortName.apply(
        lambda x : x[:x.find('购')] if '购' in x else x[:x.find('沽')])
    '''去掉嘉实基金'''
    OPsOneDay.drop(OPsOneDay[OPsOneDay.ticker.apply(lambda x:'159919' in x)].index,inplace=True)
    OPsOneDay.drop('ticker',axis=1,inplace=True)
    OPsOneDay.rename(columns={'closePrice':'p'}, inplace = True)
    '''标记出是期货期权还是指数期权'''
    checklist = ['黄金','玉米','棉花','铜','铁矿石','豆粕','甲醇','液化石油气','菜籽粕','天胶',
                 '动力煤','白糖','PTA']
    OPsOneDay['OPClass'] = OPsOneDay.biaodiwu.apply(lambda x:'FutuOP' if x in checklist else 'StkIdxOP')
    '''得到当日期权的标的合约代码'''
    varTicker_df = DataAPI.OptGet(secID=u"",optID=u"",ticker=u"",varSecID=u"",varticker=u"",contractStatus=u"",
                     exchangeCD=u"",field=[u'varTicker',u'secShortName',u"strikePrice"],pandas="1")
    '''得到当日期权的到期时间表格，并合并'''
    t_df = DataAPI.OptGet(secID=u"",optID=u"",ticker=u"",varSecID=u"",varticker=u"",contractStatus=u"",
                     exchangeCD=u"",field=[u'secID',u"expDate",u'secShortName'],pandas="1")
    
    OPsOneDay = pd.merge(OPsOneDay,varTicker_df,on='secShortName',left_index=True)
    
    '''加上每个50ETF,300ETF,沪深300期权对应的标的合约'''
    futures_contracts = DataAPI.MktMFutdGet(mainCon=u"1",contractMark=u"",contractObject=u"",tradeDate=date,
                                      startDate=u"",endDate=u"",
                                      field=u"ticker",pandas="1")
    for i in ['510050', '510300','000300']:
        futures_contracts = futures_contracts.append({'ticker':i},ignore_index=True)
    futures_contracts.rename(columns={'ticker':'varTicker'}, inplace=True)
    OPsOneDay = pd.merge(OPsOneDay,futures_contracts,on='varTicker',left_index=True)
    '''合并筛选出K'''
    OPsOneDay.rename(columns={'strikePrice':'k'}, inplace=True)
    OPsOneDay = pd.merge(OPsOneDay,t_df,on='secShortName',left_index=True)
    '''这里算一下t'''
    OPsOneDay.rename(columns={'expDate':'t'}, inplace=True)
    OPsOneDay.loc[:,'t'] = pd.to_datetime(OPsOneDay.t) - pd.to_datetime(OPsOneDay.tradeDate)
    OPsOneDay.loc[:,'t'] = OPsOneDay.loc[:,'t'].apply(lambda x: round(x.days / 365.0,6))
    '''50ETF,300ETF的t要大于0.013698,即5天'''
    #print(OPsOneDay[OPsOneDay.varTicker=='510300'])
    OPsOneDay.drop(OPsOneDay[OPsOneDay.t < 0.013698].index, inplace=True)
    #print(OPsOneDay[OPsOneDay.varTicker=='510300'])
    '''分2个部分，指数期权和期货期权'''
    FutuOPs = OPsOneDay[OPsOneDay.OPClass=='FutuOP']
    StkIdxOPs = OPsOneDay[OPsOneDay.OPClass=='StkIdxOP']
    '''获得当日所有的期货主力合约价格'''
    futuresClose_df = DataAPI.MktMFutdGet(mainCon=1,contractMark=u"",
                                          contractObject=u"",tradeDate=date,startDate=u"",
                                          endDate=u"",field=[u"secID",u"closePrice"],pandas="1")
    futuresClose_df.loc[:,'secID'] = futuresClose_df.secID.apply(lambda x:x[:x.find('.')])
    futuresClose_df.rename(columns={'secID':'varTicker'},inplace=True)
    FutuOPs.loc[FutuOPs[FutuOPs.biaodiwu=='豆粕'].index, 'varTicker'] = FutuOPs.loc[FutuOPs[FutuOPs.biaodiwu=='豆粕'].index, 'varTicker'].apply(lambda x:x.upper())
    FutuOPs.loc[FutuOPs[FutuOPs.biaodiwu.apply(lambda x:'铜' in x)].index, 'varTicker'] = FutuOPs[FutuOPs.biaodiwu.apply(lambda x:'铜' in x)].varTicker.apply(lambda x:x.upper())
    FutuOPs.loc[FutuOPs[FutuOPs.biaodiwu=='黄金'].index, 'varTicker'] = FutuOPs.loc[FutuOPs[FutuOPs.biaodiwu=='黄金'].index, 'varTicker'].apply(lambda x:x.upper())
    FutuOPs.loc[FutuOPs[FutuOPs.biaodiwu=='铁矿石'].index, 'varTicker'] = FutuOPs.loc[FutuOPs[FutuOPs.biaodiwu=='铁矿石'].index, 'varTicker'].apply(lambda x:x.upper())
    FutuOPs.loc[FutuOPs[FutuOPs.biaodiwu=='天胶'].index, 'varTicker'] = FutuOPs.loc[FutuOPs[FutuOPs.biaodiwu=='天胶'].index, 'varTicker'].apply(lambda x:x.upper())
    FutuOPs = pd.merge(FutuOPs,futuresClose_df,on='varTicker',left_index=True)
    FutuOPs.loc[:,'s'] = FutuOPs.closePrice
    FutuOPs.drop("closePrice",axis=1,inplace=True)
    
    '''获得50ETF,300ETF,000300的标的价'''
    #1按获得50ETF价格
    df50close = DataAPI.MktFunddGet(ticker=u'510050',tradeDate=date,field=
                                    [u"secID",u"secShortName",u"closePrice"],pandas="1").closePrice.values[0]
    StkIdxOPs.loc[StkIdxOPs.biaodiwu=='50ETF','s'] = df50close
    #2获得300ETF价格
    df300close = DataAPI.MktFunddGet(ticker=u'510300',tradeDate=date,field=
                                 [u"secID",u"secShortName",u"closePrice"],pandas="1").closePrice.values[0]
    StkIdxOPs.loc[StkIdxOPs.biaodiwu=='300ETF','s'] = df300close
    #print(StkIdxOPs.loc[StkIdxOPs.biaodiwu=='300ETF',:])
    #3获得000300价格
    df000300close = DataAPI.MktIdxdGet(indexID=u"",ticker=u"000300",tradeDate=date,beginDate=u"",endDate=u"",
                   exchangeCD=u"XSHE,XSHG",field=u"closeIndex",pandas="1").closeIndex.values[0]
    StkIdxOPs.loc[StkIdxOPs.biaodiwu=='沪深300','s'] = df000300close
    '''商品期权表和指数期权表合并'''
    OPsOneDay = StkIdxOPs.append(FutuOPs)
    return OPsOneDay

def FilteEqualOP(df):
    '''筛出平值期权'''
    target_df = pd.DataFrame()
    return_df = pd.DataFrame()
    FuturesHasOP= ['50ETF','黄金','玉米','棉花','铜','铁矿石','豆粕','甲醇','液化石油气',
                   '菜籽粕','天胶','动力煤','白糖','沪深300','300ETF']
    for i in FuturesHasOP:
        if df.loc[df['biaodiwu']==i,:].empty == False:
            target_df = df.loc[df['biaodiwu']==i,:]
            target_df['range'] = abs(target_df.s - target_df.k)
            target_df = target_df[target_df.range==min(target_df['range'])]
            target_df['type'] = target_df.secShortName.apply(lambda x: 1 if '购' in x else -1)
            target_df.sort_values(by='t',inplace=True)
            call_df = target_df[target_df.type==1].head(1)
            put_df = target_df[target_df.type==-1].head(1)
            return_df = return_df.append(call_df,ignore_index=True)
            return_df = return_df.append(put_df,ignore_index=True)
            return_df.drop(['range','OPClass','varTicker'],axis=1,inplace=True)
            order = ['secShortName', 'k', 's', 'p', 't','turnoverVol','type','biaodiwu','tradeDate']
            return_df = return_df[order]
    return return_df


def bs_formula(optionType, strike, spot,maturity, quotePrice):
    '''套用BS推导公式函数，固定参数r=%4，红利=0'''
    riskFree=0.04
    dividend=0
    return BSMImpliedVolatity(optionType, strike, spot, riskFree, dividend, maturity, quotePrice)

def CalcWitedIV(df,date):
    '''计算加权iv'''
    biaodiwu_set = set(df.biaodiwu.tolist())
    result_df = pd.DataFrame(index=[date])
    for i in biaodiwu_set:
        temp = df[df.biaodiwu == i]
        iv = []
        weight = []
        for index,row in temp.iterrows():
            iv_value = bs_formula(row['type'], row['k'], row['s'], row['t'], row['p'])
            iv_value = iv_value.values[0][0]
            if iv_value is None:
                print( row['secShortName'], date,row['type'], row['k'], row['s'], row['t'], row['p'] )
                iv.append(0)
                weight.append(1)
            else:
                iv.append(iv_value)
                weight.append(df.loc[index,'turnoverVol'])
        if len(weight) < 2 :
            print( row['secShortName'], date,row['type'], row['k'], row['s'], row['t'], row['p'] ,'长度少')
            weight.append(0)
        if len(iv) < 2:
            print( row['secShortName'], date,row['type'], row['k'], row['s'], row['t'], row['p'] ,'长度少')
            iv.append(0)
        wiv = (iv[0]*weight[0]+iv[1]*weight[1])/sum(weight)
        result_df[i] = wiv
        result_df['date'] = date
        #print(result_df)
    return result_df

def getTradeCalender():
    latest_date = pd.to_datetime(datetime.datetime.now()).strftime('%Y%m%d')
    
    TradeCalender = DataAPI.TradeCalGet(exchangeCD=u"XSHG",beginDate=u"20150401",endDate=u"%s"%latest_date,
                                    isOpen=u"1",field=u"calendarDate",pandas="1"
                                   ).rename(columns={'calendarDate':'date'})
    TradeCalender.loc[:,'date'] = TradeCalender.date.apply(lambda x:x.replace('-',''))
    return TradeCalender
```

```python
'''修改日期参数。这里参数为2，意思取出前推2日数据。参数1000，取出前推1000日数据'''
TradeCalender = getTradeCalender().tail(2)

FinalDF = pd.DataFrame()
ind = 0
for i , r in TradeCalender.iterrows(): 
    if ind == 3:
        ind = 0
        clear_output()
    date = r['date']
    ind += 1
    df = GetAllOPOneDay(date)
    df = FilteEqualOP(df)
    resultOneDay = CalcWitedIV(df,date)
    FinalDF = FinalDF.append(resultOneDay,ignore_index=True) 
'''这里用了打印，可以不打印自行存储FinalDF'''
print(FinalDF.tail(3))
```
