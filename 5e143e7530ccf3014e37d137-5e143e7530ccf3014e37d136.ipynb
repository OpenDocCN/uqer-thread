import requests #导入库
from bs4 import BeautifulSoup
import pandas as pd


def downLoad(url):
	headx={
	       'User-Agent': 'Mozilla/5.0 (Windows NT 6.1; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/69.0.3497.100 Safari/537.36',
	       'hexin-v': 'AsEorlDRewTA8Jcwi3Lje-fi0AbY7jZ1X2TZ2CMXv3Im6Otwaz5FsO-y6cOw',
	       } 
	res=requests.get(url,headers=headx)
	res.encoding=res.encoding#更改编码 另外值：ISO-8859-1
	res.close()
	if res.status_code==200:
		return res.content
		return None

def downLoadTesila(path='tesila.csv'):
	"""
	实时下载特斯拉分类数据
	"""
	url="http://q.10jqka.com.cn/gn/detail/field/264648/order/desc/page/{0}/ajax/1/code/301121"
	urls=[url.format(r) for r in range(1,12)]

	a=[]
	for url in urls:
		res=downLoad(url)
		if res is not None:
			soup=BeautifulSoup(res, "lxml")
			data=soup.select('body > table > tbody > tr')
			for item in data:
			    a.append([td.get_text() for td in item.find_all('td')])

	cols=['code','名称','现价','涨跌幅(%)','涨跌','涨速(%)','换手(%)','量比','振幅(%)','成交额','流通股','流通市值','市盈率']
	data=pd.DataFrame(a)
	data=data.iloc[:,1:14]
	data.columns=cols
	data.set_index("code",drop=True)
	data.to_csv(path)
	return data

def readTesila(path='tesila.csv'):
	data=pd.read_csv(path, header=0, index_col=0,dtype={'code':str})
	return data
