# 【统计套利】相关系数矩阵计算

以下代码可以对计算一段时间内一定的股票池的相关系数矩阵，并挑出正相关和负相关最强的若干只股票，其结果可以用于【配对交易】策略，见 https://uqer.datayes.com/community/share/559c85baf9f06cb5614f190d

注意：
1. 计算相关系数矩阵的起止日期和回测的起止日期要注意先后顺序，防止使用未来信息
2. 如果股票数量太多会导致计算时间过长



```python
from heapq import nlargest, nsmallest
import numpy as np

# 计算相关性矩阵行情起止时间
start = '20120101'
end = '20130101'

# 股票池，如果是 n 只则相关系数矩阵含有 n**2 个元素，有效相关系数有 n*(n-1)/2 个
# ！！如果设成全A股需要运行很久
stocklist = set_universe('ZZ500', start)

# 获取行情
stlen = len(DataAPI.MktIdxdGet(indexID = '000300.ZICN', beginDate = start, endDate = end, field='closeIndex')) # 标准长度
allclsprx = []
for i,s in enumerate(stocklist):
    z = DataAPI.MktEqudAdjGet(secID = s, beginDate = start, endDate = end, field='closePrice')
    if len(z) != stlen:
        stocklist.remove(s)
    else:
        allclsprx.append(z['closePrice'].tolist())
idxmap = {i:s for i,s in enumerate(stocklist)}
allclsprx = np.matrix(allclsprx)

# 计算相关系数矩阵，获取相关系数列表，排序
covmat = np.corrcoef(allclsprx, y=None, rowvar=1, bias=0, ddof=None)
d = len(covmat)
covlist = []
for i in range(d-1):
    for j in range(i+1, d):
        covlist.append((covmat[i, j], i, j))

# 挑出负相关系数与正相关系数最大的各10只股票
neglist = nsmallest(10, covlist, key=lambda x:x[0])
neglist = [map(idxmap.get, p[1:])+[p[0]] for p in neglist]
poslist = nlargest(10, covlist, key=lambda x:x[0])
poslist = [map(idxmap.get, p[1:])+[p[0]] for p in poslist]
```
