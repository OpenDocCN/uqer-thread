# 因子分析
对各因子进行IC分析

```python
monthend = monthend[1:-1] # 第一个月和最后一个月不计算
```

```python
# 1.因子数据获取

def get_factor(factor_name):
    field = ["ticker", "tradeDate", factor_name]
    factor = pd.DataFrame([])
    for i in range(len(monthend)):
        # 当月收益率数据
        data_1month = data[data['tradeDate'] == monthend[i]]

        # 当月股票池
        stocks = data_1month['tickerEqu'].unique()

        # 当月因子值
        nature_factor = DataAPI.MktStockFactorsOneDayGet(tradeDate=monthend[i],secID=u"",ticker=stocks,field=field,pandas="1")
        nature_factor = nature_factor.set_index(['tradeDate', 'ticker'])
        
        # 去极值
        tmp_factor = outlier(nature_factor) 
        # 标准化
        tmp_factor = z_score_standardizer(tmp_factor)
        
        # 合并原始因子值与处理后的因子值
        tmp_factor = nature_factor.merge(tmp_factor, left_index=True, right_index=True, suffixes=['_nature', ''])
        factor = factor.append(tmp_factor) # 原始因子值
    return factor

# 2.因子数据处理

# 2.1.去极值
def outlier(data_df, method='MAD', n=3): # 3Std与MAD方法，默认为MAD方法
    if method not in ['3Std', 'MAD']:
        raise BaseException(' "method" must in ["3Std", "MAD"]')
    else:
        # 1. 标准差法
        if method == '3Std':
            factor_mean = data_df.mean()
            factor_std = data_df.std()
            upper_limit = factor_mean + n * factor_std
            lower_limit = factor_mean - n * factor_std
            data_df = data_df.clip_upper(upper_limit, axis='columns')
            data_df = data_df.clip_lower(lower_limit, axis='columns')
        # 1. MAD方法
        else : 
            factor_mean = data_df.mean()
            factor_mad = data_df.mad()
            upper_limit = factor_mean + n * factor_mad
            lower_limit = factor_mean - n * factor_mad
            data_df = data_df.clip_upper(upper_limit, axis='columns')
            data_df = data_df.clip_lower(lower_limit, axis='columns')
    return data_df


# 2.2.标准化（z-score法）
def z_score_standardizer(data_df):
    # z-socre 法
    factor_mean = data_df.mean()
    factor_std = data_df.std()
    data_df = data_df.sub(factor_mean)
    data_df = data_df.div(factor_std)
    return data_df

# 2.3.因子分析
    # 计算IC
def cal_IC(ret, factor, method='rankIC'): # IC与rankIC，默认为rankIC
    if method not in ['rankIC', 'IC']:
        raise BaseException(' "method" must in ["rankIC", "IC"]')
    else:
        # ret_rank = ret.rank()
        # factor_rank = factor.rank()
        # if method == 'rankIC':
        #     coef = ret_rank.corr(factor_rank)
        # elif method == 'IC':
        #     coef = ret.corr(factor)
        # return coef
        if method == 'rankIC':
            coef, p_value = spearmanr(ret, factor)
        elif method == 'IC':
            coef, p_value = pearsonr(ret, factor)
        return (coef, p_value)
    
    # 计算IC序列
def get_IC_series(data_df, ret, factor, method='rankIC'):
    IC_series = []
    ret_factor = zip(data_df.groupby('tradeDate')[ret], data_df.groupby('tradeDate')[factor])
    for x, y in ret_factor:
        IC_series.append(cal_IC(x[1], y[1], method=method))
    IC_series = pd.DataFrame(IC_series, columns=[factor, 'P_value'], index=sorted(data_df['tradeDate'].unique()))
    return IC_series
    
    
    # 统计ICIR
def cal_ICIR(IC_series):
    IC_mean = IC_series.mean()
    IC_std = IC_series.std()
    ICIR = IC_mean / IC_std
    IC_stats = pd.DataFrame({'IC_mean':IC_mean, 'IC_std':IC_std, 'ICIR':ICIR})
    return IC_stats
```

```python
# 获取各组净值，多空净值
def get_net(data, factor):
    
    # 每一天根据因子值从小到大把所有可转债分为3组
    bin = ['bottom', 'median', 'top']
    for i in data.groupby('tradeDate'):
        data.loc[i[1].index, 'group'] = pd.qcut(i[1][factor], 3, bin)
        
    # 可转债

    # 分别统计每组的可转债的平均收益率
    symbol_net = 'NextnetBond'
    ret = data.groupby(['tradeDate', 'group'])[symbol_net].mean()

    # 等权指数收益率
    Bond_ret = data.groupby('tradeDate')['NextnetBond'].mean()
    # 等权指数净值
    Bond_net = pd.DataFrame((Bond_ret+1).cumprod().values, index=Bond_ret.index.get_level_values('tradeDate'), columns=['Bond_net']) # 累计收益率
    
    # top, median，bottom组平均收益率
    top_net = ret[ret.index.get_level_values('group') == 'top'] 
    median_net = ret[ret.index.get_level_values('group') == 'median'] 
    bottom_net = ret[ret.index.get_level_values('group') == 'bottom'] 
    # top,median,bottom净值
    net1 = pd.DataFrame((top_net+1).cumprod().values, index=top_net.index.get_level_values('tradeDate'), columns=['top_net']) # 累计收益率
    net2 = pd.DataFrame((median_net+1).cumprod().values, index=median_net.index.get_level_values('tradeDate'), columns=['median_net']) # 累计收益率
    net3 = pd.DataFrame((bottom_net+1).cumprod().values, index=bottom_net.index.get_level_values('tradeDate'), columns=['bottom_net']) # 累计收益率

    # 多空收益率
    top_bottom_return = pd.DataFrame(top_net.values - bottom_net.values, index=top_net.index.get_level_values('tradeDate'), columns=['top_bottom_return'])
    # 多空净值
    net4 = pd.DataFrame((top_bottom_return+1).cumprod().values, index=top_bottom_return.index.get_level_values('tradeDate'), columns=['top_bottom_net']) # 累计收益率

    # net_result = pd.concat([return_Equity_market, net_Equity_market, return_Bond_market, net_Bond_market, net1, net2, net3, top_bottom_return, net4], axis=1)
    net_result = pd.concat([Bond_net, net1, net2, net3, top_bottom_return, net4], axis=1)
    # 
    return net_result
```

```python
# 获取各组净值序列（可转债等权指数，top组，midian组，bottom组，top-bottom组）
net_result = pd.DataFrame([])
for i in range(len(factor_name)):
    
    # 获取各组，多空收益率
    tmp_net_result = get_net(data, factor_name[i])
    
    # 输出图片结果
    tmp_net_result[['Bond_net', 'top_net', 'median_net', 'bottom_net', 'top_bottom_net']].plot(figsize=(15, 8), title=factor_name[i])
    
    tmp_net_result['label'] = factor_name[i]
    net_result = net_result.append(tmp_net_result)
```

# 统计可转债（正股）等权指数，各因子多空策略的 （年化收益率，年化波动率，夏普比，最大回撤） 

```python
'''
year = 策略运行的交易月/12
1.计算年化收益率
采用了复利累积的年化方式计算得到的年化收益:
年化收益率 = (1 + R)^(1/year) - 1 
2.计算年化波动率
年化波动率 = sqrt( ((r_t - r_mean)^2).sum()/(month-1) *12 )
r_t：t期收益率
r_mean：样本期收益率均值
month：投资总月份
3.夏普比
sharpe = [策略年化收益率 - 无风险利率（此处未提取数据，暂时忽略] / 策略年化波动率
3.最大回撤
'''


# 计算年化收益率
# 输入最后净值，累计投资年数
def get_annual_return(net, t):
    R = (net-1.0) / 1.0 # 累计收益率
    annual_return = np.power(R+1.0, 1.0/t) - 1
    return annual_return

# 计算年化波动率
# 输入算过net_result后的data(有分组)
def get_annual_volatility(ret ,month):
    annual_volatility = np.sqrt( ( ( ret - ret.mean() ) ** 2 ).sum() / (month -1) * 12 )
    return annual_volatility

# 计算sharpe比
def get_sharpe(annual_return, annual_volatility):
    sharpe = annual_return / annual_volatility
    return sharpe

# 计算最大回撤
def get_MaxDrawdown(return_list):
    '''最大回撤率'''
    i = np.argmax((np.maximum.accumulate(return_list) - return_list) / np.maximum.accumulate(return_list))  # 结束位置
    if i == 0:
        return 0
    j = np.argmax(return_list[:i])  # 开始位置
    MaxDrawdown = (return_list[j] - return_list[i]) / (return_list[j])
    return MaxDrawdown
```

```python
# 年化收益率，年化波动率，夏普比，最大回撤 指标统计

month = len(monthend)
t = month/12 # 累计投资年数
net = []
annual_return = []
annual_volatility = []
sharpe = []
maxdown = []

Equity_Bond = ['Equity', 'Bond']    

# 1.统计等权指数
for i in range(len(Equity_Bond)):
    if Equity_Bond[i] == 'Equity':
        tmp_net = Equity_net[1:-1]
        tmp_ret = Equity_ret[1:-1]
    else:
        tmp_net = Bond_net[1:-1]
        tmp_ret = Equity_ret[1:-1]
        
        
    # 计算年化收益率
    net.append(tmp_net[-1]) # 最后一日净值
    tmp_annual_return = get_annual_return(tmp_net[-1], t)
    annual_return.append(tmp_annual_return)
    
    # 计算年化波动率
    tmp_annual_volatility = get_annual_volatility(tmp_ret, month)
    annual_volatility.append(tmp_annual_volatility)
    
    # 计算sharpe
    tmp_sharpe = get_sharpe(tmp_annual_return, tmp_annual_volatility)
    sharpe.append(tmp_sharpe)
    
    # 计算最大回撤
    tmp_maxdown = get_MaxDrawdown(tmp_net)
    maxdown.append(tmp_maxdown)
    
# 2.统计因子多空策略
for i in range(len(factor_name)):
    # 计算年化收益率
    tmp_net = net_result[net_result['label'] == factor_name[i]]['top_bottom_net']
    net.append(tmp_net[-1]) # 最后一日净值
    tmp_annual_return = get_annual_return(tmp_net[-1], t)
    annual_return.append(tmp_annual_return)
    
    # 计算年化波动率
    tmp_ret = net_result[net_result['label'] == factor_name[i]]['top_bottom_return']
    tmp_annual_volatility = get_annual_volatility(tmp_ret, month)
    annual_volatility.append(tmp_annual_volatility)
    
    # 计算sharpe
    tmp_sharpe = get_sharpe(tmp_annual_return, tmp_annual_volatility)
    sharpe.append(tmp_sharpe)
    
    # 计算最大回撤
    tmp_maxdown = get_MaxDrawdown(tmp_net)
    maxdown.append(tmp_maxdown)
    


    
Equity_Bond.extend(factor_name)    
result = pd.DataFrame({
    '最终净值':net, '累计投资年数':[t]*len(Equity_Bond), '年化收益率':annual_return, 
    '年化波动率':annual_volatility, 'sharpe':sharpe, '最大回撤':maxdown
}, index=Equity_Bond)

```

# 因子回归

$r_{t+1} = factor_t × X_t + u_t$
r_{t+1}：t+1期个股超额收益率（以等权指数收益率作为基准）  
factor_t：t期因子暴露度  
X_t：t期因子收益率（回归拟合）  
u_t：残差收益率  


```python
Bond_ret
```

```python
# 单因子回归
def single_factor_regress(factor, symbol='Bond'):
    coef = []
    t_value = []
    # 每一期进行横截面回归
    for i in range(len(monthend)):
        tmp_data = data[data['tradeDate'] == monthend[i]]
        if symbol == 'Bond':
            X = tmp_data['NextnetBond'] - Bond_ret[1:-1][i] # 超额收益率
        elif symbol == 'Equity':
            X = tmp_data['NextnetEqu'] - Equity_ret[1:-1][i] # 超额收益率
        else:
            raise BaseException(' "symbol" must in ["Bond", "Equity"]')
            
        Y = tmp_data[factor]
        model = sm.OLS(X, Y)
        fit = model.fit()

        coef.append(fit.params[0]) # t期因子收益率
        t_value.append(fit.tvalues[0]) # t期因子收益率的t值
        
    reg_result = pd.DataFrame({'coef':coef,'t_value':t_value}, index=monthend)
    reg_result['abs_t_value'] = abs(reg_result['t_value'])
    reg_result['factor'] = factor
    reg_result['symbol'] = symbol

    return reg_result
```

```python
symbol_name = ['Bond', 'Equity']
result_single_factor_regress = pd.DataFrame([])
for i in range(len(factor_name)):
    for j in range(len(symbol_name)):
        result_single_factor_regress = result_single_factor_regress.append(single_factor_regress(factor_name[i], symbol=symbol_name[j]))
```
