__算法介绍__
K-means 算法是一种聚类算法，或者用机器学习的语言说是一种无监督学习。这类算法不需要我们预先知道结果的训练样本，可以直接对我们需要处理的数据进行分析。所谓聚类问题就是，给定一组数据$\{x^{(1)},\cdots,x^{(m)}\}$，我们希望将这些数据按它们之间的联系紧密程度分类。这里 $x^{(i)}\in\mathbb{R}^n$，我们没有像通常的Logistic Regression或者SVM分类算法中的标签 $y^{(i)}$，因此这是一种无监督学习。
算法可表述如下：
1.随机生成类中心点$\mu_1,\mu_2,\cdots,\mu_k$.
2.重复以下过程直至收敛：{
	对每个$i$,计算：
    	$c^{(i)}=\arg\min\limits_j||x^{(i)}-\mu_j||^2$
    对每个$j$,计算：
    	$\mu_j=\frac{\sum_{i=1}^{m}\mathbb{I}_{\{c^{(i)}=j\}}x^{(i)}}{\sum_{i=1}^{m}\mathbb{I}_{\{c^{(i)}=j\}}}$
    }
其中$k$是我们想分的类数，$\mu_j$是我们当前对聚类中心的估算，$\mathbb{I}_{A}$是示性函数。
第一步我们随机生成了$k$个聚类中心，在算法的第二个步骤中包含一个循环，这个循环分为两步：（i）对每个数据点，找出它离哪个聚类中心最近并标记；（ii）对我们标好类的数据计算其中心，把计算好的结果做为我们对聚类中心新的估计。重复以上过程直至收敛。

__例子__
学习完了这个算法之后，我就跃跃欲试，可惜一时找不到可供实验的数据，我就自己生成了一些。以下考虑了k-means算法的简单情形，我们对二维欧式空间的数据集进行聚类。

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pickle
def Data_Generator(alpha):
    x1=np.random.rand(35)
    x2=np.random.rand(35)
    y1=1+np.random.rand(40)+alpha
    y2=np.random.rand(40)
    z1=np.random.rand(47)
    z2=1+np.random.rand(47)+alpha
    w1=1+np.random.rand(61)+alpha
    w2=1+np.random.rand(61)+alpha
    x=np.hstack((x1,y1,z1,w1))
    y=np.hstack((x2,y2,z2,w2))
    return x,y
#out=open('kmeans_data.pkl','wb')
#pickle.dump([x,y],out)
#out.close()
#data=open('kmeans_data.pkl','rb')
#l=pickle.load(data)
#x=l[0]
#y=l[1]
#plt.scatter(x, y)
#plt.show()
class Kmeans():
    def __init__(self,k,iters,data):
        self.clusters=k
        self.iters=iters
        self.x=data[0]
        self.y=data[1]
        np.random.seed(0)
        self.centers=np.random.rand(k,2)
    def k_means(self):
        for it in range(self.iters):
            classified_data={}
            for j in range(self.clusters):
                classified_data['cluster_'+str(j)]=[]
            for (a,b) in zip(self.x,self.y):
                distance=[((a - c) ** 2 + (b - d) ** 2) ** 0.5 for (c,d) in zip(self.centers.T[0],self.centers.T[1])]
                index=distance.index(min(distance))
                classified_data['cluster_'+str(index)].append([a,b])
            for k in range(self.clusters):
                l0=[classified_data['cluster_'+str(k)][i][0] for i in range(len(classified_data['cluster_'+str(k)]))]
                self.centers[k][0]=sum(l0)/len(l0)
                l1=[classified_data['cluster_'+str(k)][i][1] for i in range(len(classified_data['cluster_'+str(k)]))]
                self.centers[k][1]=sum(l1)/len(l1)
        return classified_data
#例子1，自己生成的数据，0.1为噪音控制，分为4类
data=Data_Generator(0.1)
example1=Kmeans(4,200,data)
classified_data=example1.k_means()
colors = ['r', 'g', 'y', 'k']
x = {}
y = {}
for i in range(example1.clusters):
    x[str(i)] = np.array(classified_data['cluster_' + str(i)]).T[0]
    y[str(i)] = np.array(classified_data['cluster_' + str(i)]).T[1]
    plt.scatter(x[str(i)], y[str(i)], color=colors[i])
plt.show()
#例子2.下面是网上找的数据，分为2类
data = [[6.9, 3.1, 1.0], [5.0, 3.5, 0.0], [5.0, 2.3, 1.0], [4.6, 3.4, 0.0], [5.5, 2.4, 1.0], [5.5, 3.5, 0.0], [5.2, 4.1, 0.0], [4.7, 3.2, 0.0], [5.4, 3.9, 0.0], [5.6, 3.0, 1.0], [6.1, 2.8, 1.0], [5.8, 2.7, 1.0], [6.2, 2.2, 1.0], [5.0, 3.4, 0.0], [5.7, 3.8, 0.0], [6.3, 2.5, 1.0], [6.0, 3.4, 1.0], [4.9, 3.1, 0.0], [4.9, 3.0, 0.0], [5.6, 2.7, 1.0], [5.7, 2.8, 1.0], [5.0, 3.2, 0.0], [5.9, 3.0, 1.0], [6.1, 2.8, 1.0], [5.1, 3.5, 0.0], [4.6, 3.2, 0.0], [5.1, 2.5, 1.0], [5.1, 3.7, 0.0], [5.4, 3.7, 0.0], [5.3, 3.7, 0.0], [4.8, 3.0, 0.0], [4.3, 3.0, 0.0], [4.9, 2.4, 1.0], [5.5, 2.5, 1.0], [5.8, 4.0, 0.0], [5.1, 3.8, 0.0], [5.7, 2.8, 1.0], [5.1, 3.8, 0.0], [5.0, 3.0, 0.0], [4.4, 3.0, 0.0], [5.4, 3.4, 0.0], [5.7, 2.6, 1.0], [6.4, 3.2, 1.0], [5.2, 3.5, 0.0], [4.8, 3.4, 0.0], [6.0, 2.7, 1.0], [5.6, 3.0, 1.0], [5.1, 3.3, 0.0], [5.4, 3.9, 0.0], [6.8, 2.8, 1.0], [5.9, 3.2, 1.0], [4.5, 2.3, 0.0], [6.1, 3.0, 1.0], [4.4, 2.9, 0.0], [5.4, 3.4, 0.0], [5.5, 2.4, 1.0], [5.6, 2.9, 1.0], [5.5, 2.6, 1.0], [5.7, 4.4, 0.0], [5.0, 3.6, 0.0], [5.0, 3.4, 0.0], [4.6, 3.6, 0.0], [5.7, 2.9, 1.0], [5.0, 2.0, 1.0], [6.3, 2.3, 1.0], [7.0, 3.2, 1.0], [4.6, 3.1, 0.0], [5.0, 3.3, 0.0], [5.1, 3.8, 0.0], [4.9, 3.1, 0.0], [4.7, 3.2, 0.0], [5.5, 2.3, 1.0], [6.7, 3.1, 1.0], [6.5, 2.8, 1.0], [6.7, 3.0, 1.0], [4.4, 3.2, 0.0], [6.6, 2.9, 1.0], [5.0, 3.5, 0.0], [6.4, 2.9, 1.0], [5.1, 3.5, 0.0], [6.3, 3.3, 1.0], [5.4, 3.0, 1.0], [6.2, 2.9, 1.0], [5.1, 3.4, 0.0], [5.2, 3.4, 0.0], [4.8, 3.0, 0.0], [6.0, 2.9, 1.0], [5.5, 4.2, 0.0], [4.8, 3.4, 0.0], [5.8, 2.6, 1.0], [5.8, 2.7, 1.0], [6.7, 3.1, 1.0], [4.8, 3.1, 0.0], [5.7, 3.0, 1.0], [6.1, 2.9, 1.0], [6.0, 2.2, 1.0], [6.6, 3.0, 1.0], [4.9, 3.1, 0.0], [5.6, 2.5, 1.0], [5.2, 2.7, 1.0]]
data=np.array(data)
data=data.T
data=(data[:-1][0],data[:-1][1])
example2=Kmeans(2,200,data)
classified_data=example2.k_means()
colors = ['r', 'g', 'y', 'k']
x = {}
y = {}
for i in range(example2.clusters):
    x[str(i)] = np.array(classified_data['cluster_' + str(i)]).T[0]
    y[str(i)] = np.array(classified_data['cluster_' + str(i)]).T[1]
    plt.scatter(x[str(i)], y[str(i)], color=colors[i])
plt.show()
```
