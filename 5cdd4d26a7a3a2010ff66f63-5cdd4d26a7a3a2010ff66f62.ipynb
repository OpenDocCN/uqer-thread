# 筛选有效因子

```python
import pandas as pd #导入pandas工具包
import numpy as np #导入numpy工具包

# 定义spearman秩相关
def spearman(df):
    return df.corr("spearman")
data=DataAPI.TradeCalGet(exchangeCD=u"XSHG",beginDate=u"20150101",endDate=u"20171010",field=['calendarDate','isWeekEnd'],pandas="1")
data = data[data['isWeekEnd'] == 1]
date_list = data['calendarDate'].values.tolist()

coormaxtri=[]
for i in range(len(date_list)-1):
    hscyb_secID = DataAPI.IdxConsGet(secID=u"",ticker=["000300","399006"],isNew=u"",intoDate=date_list[i],field=["consID"],pandas="1")
    Factors = DataAPI.MktStockFactorsOneDayGet(tradeDate=date_list[i],secID=hscyb_secID["consID"],field=u"",pandas="1")
    chgPct = DataAPI.MktEquwGet(secID=hscyb_secID["consID"],ticker=u"",weekEndDate=date_list[i+1],isOpen=u"1",field=["secID","chgPct"],pandas="1")
    result = pd.merge(Factors, chgPct, on='secID')
    del result["ticker"];del result["tradeDate"]
    #result=result.fillna(result.mean()).reset_index(drop = True)#清洗数据
    result=result.dropna().reset_index(drop = True)#清洗数据
    result.set_index(['secID'],inplace=True)
    coormaxtrii=spearman(result)
    coormaxtri.append(coormaxtrii)
    coormaxtri_avr = sum(coormaxtri)/(len(date_list)-1)
```

```python
coormaxtri_avr.to_csv('coormaxtri_avr1.csv')
```

```python
import pandas as pd #导入pandas工具包
#  将数据写入新文件
coormaxtri_avr=pd.read_csv('coormaxtri_avr1.csv')
coormaxtri_avr.set_index(['Unnamed: 0'],inplace=True)
coormaxtri_avr=coormaxtri_avr.dropna(axis=0,how="all")
coormaxtri_avr=coormaxtri_avr.dropna(axis=1,how="all")
coormaxtri_avr["chgPct"]=abs(coormaxtri_avr["chgPct"])

coormaxtri_avr = coormaxtri_avr.sort_index(by = "chgPct",ascending = False)
coorfactorcode = list(coormaxtri_avr[0:61]["chgPct"].index)
#coorfactorcode = coormaxtri_avr[coormaxtri_avr["chgPct"]>0.02].index
coorfactorcode = coorfactorcode[1:]
```

```python
#筛选出有效因子与有效因子的相关系数
coorfactorcodelist = coorfactorcode[:]
beixuanfactor_maxtri = coormaxtri_avr.ix[coorfactorcodelist][coorfactorcodelist]
beixuanfactor_maxtri
```

```python
coorfactorcodeset = set(coorfactorcodelist)
for i in range(beixuanfactor_maxtri.shape[0]):
    for j in range(beixuanfactor_maxtri.shape[1]):
        if i>=j:
            continue
        else:
            if beixuanfactor_maxtri.iloc[i,j] >= 0.8:
                if coormaxtri_avr.loc[beixuanfactor_maxtri.index[i],"chgPct"] >= coormaxtri_avr.loc[beixuanfactor_maxtri.index[j],"chgPct"]:
                    coorfactorcodeset.discard(beixuanfactor_maxtri.index[j])
                else:
                    coorfactorcodeset.discard(beixuanfactor_maxtri.index[i])
coorfactorcodelist_new = list(coorfactorcodeset)
coorfactorcodelist_new
```
