# A股主要指数PB、PE因子的中位数

首先我通过同花顺查询到了这些指数的ticker，然后查询这些指数的secID。

```python
import pandas as pd

# 上证50=000016；基本面50=000925；沪深300=000300；中证500=000905；深证100=399330；中小板综=399101；创业板指=399006；中证红利=000922；中证环保=000827；全指医药=000991；全指消费=000990；全指消费=000993

# 由ticker获取指数的额secID
idx = DataAPI.SecIDGet(ticker=u"000016,000925,000300,000905,399330,399101,399006,000922,000827,000991,000990,000993", assetClass="IDX")
print idx
```

为了数据分析的准确性，我想获得这些指数最完整的历史数据，所以我又查询了这些指数的发布日期publishDate。发现发布时间最早的是上证50，所以在后面提取这些指数的历史数据时，我将最早时间点设置为20040102。。。但是，后来我发现通联数据能提供的最早时间是20070101以后的历史数据。。不过，也没关系啦，毕竟大部分指数都是2007您以后发布的。。

```python
# 获取指数的发布时间
DataAPI.IdxGet(idx["secID"])
```

下载所有指数的收盘价、PB、PE因子历史数据，并将其拼接成一个大的数据框。

```python
# 下载指数收盘价、PB、PE因子的历史数据
# 优矿所有指数数据都是从20070101开始具备数据的
dataall = pd.DataFrame()
for sec in idx["secID"]:
    temp = DataAPI.MktIdxFactorDateRangeGet(secID=sec,beginDate=u"20040102",endDate=u"20170120",field=u"secID,tradeDate,Close,PB1,PB2,PE1,PE2",pandas="1")
    temp = temp.set_index("secID")
    dataall = dataall.append(temp)

# 保存到本地文件
dataall.to_csv("A股指数PBPE因子历史数据.csv")

# 预览数据
print dataall.head()
print dataall.tail()
```

计算每个指数的中位数（最大值、最小值、平均值）

```python
# 计算所有指数的统计值（最大、最小、平均值、中位数）
datastas = pd.DataFrame()
for sec in idx["secID"]:
    temp = dataall.ix[sec]
    stas = temp.median(axis=0)[1:]   #忽略第一列的日期
    #Max = temp.max(axis=0)[1:]
    #Min = temp.min(axis=0)[1:]
    #Mean = temp.mean(axis=0)[1:]
    stas.name = sec
    datastas = pd.concat([datastas, stas], axis=1)

# 保存到本地文件
datastas.to_csv("A股指数PBPE因子中位数.csv")
#datastas.to_csv("A股指数PBPE因子最大值.csv")
#datastas.to_csv("A股指数PBPE因子最小值.csv")
#datastas.to_csv("A股指数PBPE因子平均值.csv")

# 预览数据
print datastas
```
