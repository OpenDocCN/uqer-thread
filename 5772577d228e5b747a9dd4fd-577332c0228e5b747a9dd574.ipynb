## 期货折价与现货溢价  
### <a>1. 期货合约的一般无套利定价</a>  

  * 一般来说，在无套利的情形下，如果标的物在期货合约过程中不提供或消耗现金流，期货价格\(F\)与现货价格\(S\)应该满足：\(F=Se^{rT}\)，其中\(r\)是连续复利，\(T\)是到交割日期的剩余时长。   
  
  * 其实这也容易理解，如果\(F<Se^{rT}\)，可以在现货市场做空，得到\(S\)现金，以无风险利率\(r\)投资，并签订远期（期货）条约，以\(F\)在\(T\)时间后买入标的物。在时刻\(T\)，投资者获得投资收益\(Se^{rT}\)，用\(F\)买入标的物平了现货市场空仓，套利\(Se^{rT}-T\). 如果\(F>Se^{rT}\)，则也可以反向操作套利。
  * 如果标的物能够产生收入（现值为\(I\)），有储藏成本（现值为\(U\)），则上述关系变为\(F=(S+U-I)e^{rT}\).  
  
  * 然而，利用优矿的期货、现货和股指等数据，可以发现中国市场上的奇怪现象——现货价格常常高于期货价格。

* 为了观察、分析这些现象，先引入下列工具包：

```python
from __future__ import division
import datetime
import time
import pandas as pd
import string
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
matplotlib.style.use('ggplot')
import seaborn as sns
```

* 为了统一和修改方便，在这里定义一些全局变量：

```python
begDate = u"20150101"     # 研究开始日期
fsize = (12, 12*0.618)         # 图画大小
```

* 定义获取连续期货价格的函数 getFutuCon()：

```python
def toDateTime(argStr = u"2012-01-01"):
    return datetime.date(string.atoi(argStr[0:4]), string.atoi(argStr[5:7]), string.atoi(argStr[8:10]))
def getFutuCon(bDate = u"20120101", eDate = u"", conObj = u"AU", period = 1):
    histPrice = {}
    if eDate == "":
        eDate = time.strftime("%Y%m%d",  time.localtime(time.time()))
    calendar = DataAPI.TradeCalGet(exchangeCD = u"XSHE", beginDate = bDate, endDate = eDate, field = u"calendarDate,isOpen", pandas = "1")
    calArray = calendar[calendar['isOpen'] == 1]['calendarDate'].values
    for date in calArray:
        year = string.atoi(date[2:4]);month = string.atoi(date[5:7])
        year = (year if (month+period<= 12) else year+1)
        month = (month+period if (month+period<= 12) else month+period-12)
        year = str(year)
        month = ('0'+str(month) if(month<10) else str(month))
        tker = conObj+year+month
        priceVal = DataAPI.MktFutdGet(tradeDate = toDateTime(date), ticker = tker, field = u"closePrice")
        histPrice[toDateTime(date)] = (np.NaN if len(priceVal) == 0 else priceVal)
    return pd.DataFrame(histPrice, index = [conObj+'_Continuous_'+str(period)]).T
```

### <a>2. 股指期货折价</a>    
* 从图中可以看见，2015年以来，连续1月的期货价格在大部分时候都是高于现货股指的。如果上文所述的套利过程能够实现，这个现象显然是不能持续的。  

* 说这一现象是因为股指没有卖空机制，恐怕解释得也不完全。因为期货价格低于股指时，股指ETF的持有者也可以卖出持有的ETF，转而持有更便宜的期货合约，从而期货需求量上升，期货价格低于现货的现象消失。然而现实并不是无套利理论预测的样子，不知道大家有没有合理的解释呢？ 

```python
histIdx = DataAPI.MktIdxdGet(ticker = '000300', beginDate = begDate, field = 'closeIndex,tradeDate').rename_axis(dict(closeIndex = 'HS300'), axis = 1)
histIdx['tradeDate'] = map(toDateTime, histIdx['tradeDate'])
histIdx = histIdx.set_index(keys = 'tradeDate')
histIdx = histIdx.join(getFutuCon(begDate, conObj = 'IF', period = 1))
histIdx.plot(figsize = fsize)
```

### <a>3. 黄金期货折价</a>  

* 下方第一张图显示黄金期货连续1月合约、连续两月合约与现货市场T+D交割黄金价格的关系，第二张图是黄金期货连续1月合约、连续两月合约价格与现货市场T+D交割黄金价格的比值。

* 可见，黄金期货合约也常常低于现货市场上的黄金价格。注意到黄金现货市场的交易是双向的，即可以做空，这一现象的持续就更加奇怪了。是什么阻止了做空现货、买入期货进行套利呢？

```python
# 获取数据
histFut1 = getFutuCon(bDate = begDate, period = 1)
histFut2 = getFutuCon(bDate = begDate, period = 2)
histSpot = DataAPI.MktSpotdGet(ticker = u"Au(T%2bD)", beginDate = begDate, field = 'closePrice,tradeDate')
histSpot['tradeDate'] = map(toDateTime, histSpot['tradeDate'])
histFut = pd.merge(histFut1, histFut2, left_index = True, right_index = True, how = 'inner')
histAu = pd.merge(histSpot, histFut,left_on = 'tradeDate', right_index = True).set_index('tradeDate').rename_axis(dict(closePrice = 'Spot'), axis = 1)
# 处理与作图
futuPremium = {}
futuPremium['month_2'] = histAu['AU_Continuous_2']/histAu['Spot']
futuPremium['month_1'] = histAu['AU_Continuous_1']/histAu['Spot']
futuPremium['base'] = histAu['Spot']/histAu['Spot']
futuPremium = pd.DataFrame(futuPremium)
# futuPremium[futuPremium>1] = np.NaN
histAu.plot(figsize = fsize)
sns.despine()
futuPremium.plot(figsize = (12,1.2*6.18))
sns.despine()
histAu[:][:1]
```

### <a>3. 小结</a>  
* 看到这，是不是有做空现货、做多期货合约套利的想法了呢？我自己今天就翻阅了各种商品期货现货交易的规则，也没有发现有什么机制阻止这样的套利行为。  

* 比如6月28日的情况：黄金现货281.43元/g，黄金1607合约280.3元/g. 期货和现货都是1000g/手，最少交易1手。期货合约交易费用10元/手，交割费0.06元/g；现货交易费4.5/10000，交割费为0，T+D延期补偿费日最高0.02%. 现货最低保证金12%，期货最低保证金10%. 考虑以上一切的时候：

> 6.28 做空一手现货，多方一手期货合约。需要付出281430\*12%=33771.6元保证金，支付手续费281430\*4.5/10000=126.64元，可得现金281430元；再支付期货合约保证金280300\*10%=28030元，手续费10元，剩余现金219491.76元。保守做法，留50000元作为保证金调整费用，避免爆仓。剩下169491.76元存入银行活期，获取0.35%的年利，可得169491.76\*0.35%\*17/360=28.01元。    
> 
> 6.28-7.15期间，支付T+D现货延期补偿费0.02%\*281430\*17=956.86元。用保留现金调整保证金数额。  
> 
> 到7月15日，取出存款，支付交割费用0.06\*1000=60元，加保证金共支付期货合约价280300元，平仓现货期货。共获得281430+28.01-126.64-10-956.86-60-280300=4.7元。
该操作实际是没有投入要求的，保守估计最终能够获得4.7元，若操作量大我想还是有利可图。
