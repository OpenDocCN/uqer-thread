```python
import seaborn
from matplotlib import pylab
import matplotlib.pyplot as plt
from matplotlib.pyplot import *
from CAL.PyCAL import *    # CAL.PyCAL中包含font
from datetime import timedelta,datetime
import time
cal=Calendar('China.SSE')
```

```python
def buylist(input_account_date,universe0):
    calDate = Date.fromDateTime(input_account_date)
    xgstart=cal.advanceDate(calDate,'-120B',BizDayConvention.Preceding)
    xgend=cal.advanceDate(calDate,'-1B',BizDayConvention.Preceding)
    buylistall=[] 
    for i in universe0:
        df=DataAPI.MktEquwAdjAfGet(secID=i,beginDate=xgstart,endDate=xgend,field=u"secID,closePrice,turnoverVol",pandas="1")
        dfprice=df.closePrice.tolist()
        dfvol=df.turnoverVol.tolist()
        #五周大于十周,出现金叉放入股票列表中 
        if sum(dfprice[-5:])/5 > sum(dfprice[-10:])/10 and sum(dfprice[-6:-1])/5 < sum(dfprice[-11:-1])/10:
            buylistall.append(i)
    return buylistall
            
```
