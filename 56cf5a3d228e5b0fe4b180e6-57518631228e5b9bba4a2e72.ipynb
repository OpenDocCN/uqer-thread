# 关于期权

这是我在优矿的第八篇文章，昨天股票大跌，我虽然小有损失，但是只是回吐了前几天的一点利润，说说期权吧。
关于什么是期权，请自行百度。
目前国内上市的期权只有上交所的50etf期权，与2015年2月上市，目前已一年，因为有50万和考试的门槛限制，开户数量不足10万，正式参与的不足3万（有交易记录），由于为上交所产品，证券公司基本都可以开户，部分期货公司也取得了开户资格，我所在的华泰证券目前大约是8元每张，据说银河期货一张只要5元，这个价格包括证交所收取的2.3元左右的税费和管理费，其余为券商自行收取的佣金。
期权将是我未来主要的投资方向，通过资料显示，卖方比买方更有优势，因此我是一个卖方，通过购沽虚值进行对冲，利用期权每天时间价值的损失获得利润。（这个是我目前的一个稳定盈利的策略，不能说多了）。
期权是一个新生事物，用好了绝对事半功倍。
